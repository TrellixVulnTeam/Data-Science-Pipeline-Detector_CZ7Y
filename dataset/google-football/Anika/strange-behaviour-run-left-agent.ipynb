## Exploring the GFootball environment

The next step is setting up the environment. To keep things simple I choose the 1 versus 1 scenario. For more information look [here](https://github.com/google-research/football/blob/master/README.md)

```python
# Set up the Environment.
from kaggle_environments import make
env = make("football", configuration={"save_video": True, 
                                      "scenario_name": "1_vs_1_easy", 
                                      "running_in_notebook": True})
```

The enviroment comes with four agents that can play football, well that is running to right, running to the left or doing nothing. One agent uses the builtin AI.

```python
env.agents
```

# Play some football

To get get a feeling of the game, lets play some football using the agents that are already present and render the game in the notebook (First I used the safari webbrowser and there was nothing to see, with chrome as a webbrowser I can see the game being played)

```python
# Let run_left agent play against eachother and see what happens
output = env.run(["run_left", "run_left"])

# show the game
env.render(mode="human", width=800, height=600)
```

One run_left agent runs towards his own goal and disappears from the field. The other player also a run_left agent runs towards the ball, gets the ball, turns around and puts the ball in his own goal!?? Lets have closer look at the run_left agent to see if we can explain this strange behaviour. 

Lets inspect the run_left_agent

```python
import inspect
import os

print(inspect.getsource(env.agents['run_left']))
```
