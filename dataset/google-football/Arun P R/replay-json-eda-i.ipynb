## Enter URL of the game of interest

```python
url = 'https://www.kaggleusercontent.com/episodes/3634538.json'
```

## Load JSON from URL

```python
import urllib, json

response = urllib.request.urlopen(url)
data = json.loads(response.read())
```

### Explore the Data

Refer [link](https://github.com/google-research/football/blob/master/gfootball/doc/observation.md) for details on how to interpret these values.

The left player is assumed to be my player. If you want data for right player, comment the existing line and uncomment the second line inside for loop.

```python
dict_obs = {}
for i in range(3002):
    dict_obs[i] = data['steps'][i][0]['observation']['players_raw'][0] # left player
    dict_obs[i]['action'] = data['steps'][i][0]['action'][0] if len(data['steps'][i][0]['action']) > 0 else -1
    #dict_obs[i] = data['steps'][i][1]['observation']['players_raw'][0] # right player
```

### Make dataframe of raw observations

```python
df = pd.DataFrame(dict_obs)
df = df.T
df
```

```python
df.describe()
```

All values are interpreted as strings by default. You can change them to appropriate type if you want mean, std, quantile values, etc.
