```python
from rake_nltk import Rake
```

```python
#r = Rake()
```

```python
r = Rake(punctuations = [')','(',',',':','),',').','.'])
```

```python
#r.extract_keywords_from_text(notes.pn_history[20].replace('(','').replace(')',''))
r.extract_keywords_from_text(notes.pn_history[20])
```

```python
phrase_df = pd.DataFrame(r.get_ranked_phrases_with_scores(), columns = ['score','phrase'])
```

```python
phrase_df.loc[phrase_df.score>3]
```
