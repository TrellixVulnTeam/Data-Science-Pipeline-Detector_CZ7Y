```python
features_per_case = features.groupby("case_num")["feature_num"].nunique().reset_index()
```

```python
# Number of unique features per case
plt.figure(figsize=(10,5))
ax = sns.barplot("case_num","feature_num", data= features_per_case)
ax.bar_label(ax.containers[0])
plt.title("Number of unique features per case")
plt.show()
```
