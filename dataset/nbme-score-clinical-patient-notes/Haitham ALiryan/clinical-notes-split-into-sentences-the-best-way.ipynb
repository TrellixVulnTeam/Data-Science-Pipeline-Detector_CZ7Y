```python
from PyRuSH import RuSH
```

```python
patient_notes = pd.read_csv("/kaggle/input/nbme-score-clinical-patient-notes/patient_notes.csv")
```

```python
patient_notes
```

```python
PATIENT_IDX = 30
rush = RuSH('conf/rush_rules.tsv')
note=patient_notes[patient_notes["pn_num"] == PATIENT_IDX]["pn_history"].iloc[0]
print("Note Before Splitting")
print(note)
print("*"*20)
sentences=rush.segToSentenceSpans(note)
print("Note After Splitting")
for sentence in sentences:
    sentence=note[sentence.begin:sentence.end]
    print(sentence)
    print("-"*20)
```
