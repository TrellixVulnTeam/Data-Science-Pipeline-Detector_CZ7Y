# Load And Merge Data

 - https://www.kaggle.com/c/nbme-score-clinical-patient-notes/data

```python
dftr = pd.read_csv("../input/nbme-score-clinical-patient-notes/train.csv")
dfft = pd.read_csv("../input/nbme-score-clinical-patient-notes/features.csv")
dfpn = pd.read_csv("../input/nbme-score-clinical-patient-notes/patient_notes.csv")
merge1 = pd.merge(dftr, dfft, how="left", on=["feature_num", "case_num"])
df = pd.merge(merge1, dfpn, how="left", on=["pn_num", "case_num"])
df.head()
```

# View Naive Start Location Distribution for Feature 0

- Does not take into consideration multiple annotations per record

```python
smydf = dftr[dftr['feature_num']==0].copy()

def get_loc1(x):
    x = x.split(' ')[0].replace('[','').replace(']','').replace("'",'').replace(' ','')
    if x=='':
        return np.nan
    else:
        return int(x)
    
smydf['loc1'] = smydf['location'].apply(get_loc1)
smydf.head()
```

```python
plt.hist(smydf.loc1.values)
plt.show() 
```

# Overall Naive Start Location Distribution 

- Does not take into consideration multiple annotations per record

```python
def get_loc1(x):
    x = x.split(' ')[0].replace('[','').replace(']','').replace("'",'').replace(' ','')
    if x=='':
        return np.nan
    else:
        return int(x)
    
df['loc1'] = df['location'].apply(get_loc1)

plt.hist(df.loc1.values)
plt.show()
```

```python
df.loc1.describe()
```

# Distribution of Total Annotations by Row

```python
def get_tot_ants(x):
    x = len(x.split(','))
    return int(x)
    
df['total_annotation_locations'] = df['location'].apply(get_tot_ants)

plt.hist(df.total_annotation_locations.values)
plt.show()
```

```python
df.total_annotation_locations.describe()
```
