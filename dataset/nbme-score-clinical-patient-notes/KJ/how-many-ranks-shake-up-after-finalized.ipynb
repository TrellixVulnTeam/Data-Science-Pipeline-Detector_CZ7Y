```python
import pandas as pd 
```

```python
pd.set_option('display.max_rows', 1500)
```

```python
# LB data before finalize

before_finalized = pd.read_csv("../input/nbmelb/LB_Before_finalized.csv")
before_finalized = before_finalized[before_finalized["TeamName"] != "sample_submission.csv"]

before_finalized = before_finalized.reset_index()
before_finalized = before_finalized.rename({"index":"before_rank"},axis = 1)
before_finalized["before_rank"] += 1
```

```python
print("LB size before finalized : ",len(before_finalized))
```

```python
# LB data after finalize
after_finalized  = pd.read_csv("../input/nbmelb-after-finalized/nbme-score-clinical-patient-notes-privateleaderboard.csv")
after_finalized  = after_finalized [after_finalized["TeamName"] != "sample_submission.csv"]
after_finalized = after_finalized.reset_index()
after_finalized = after_finalized.rename({"index":"after_rank"},axis = 1)
after_finalized["after_rank"] += 1
```

```python
print("LB size before finalized : ",len(after_finalized))
```

```python
merged = pd.merge(before_finalized.drop("Score",axis=1),after_finalized[["after_rank","TeamId","Score"]],on = "TeamId")
```

```python
merged["shake_up"] = merged["before_rank"] - merged["after_rank"]
```
