```python
fp = "../input/fast-text/yake_results.csv"
yake = pd.read_csv(fp)
pn_fp = '../input/nbme-score-clinical-patient-notes/patient_notes.csv'
pn = pd.read_csv(pn_fp)
```

```python
locations = list()
for case_num in range(10):
    pn_case = pn[pn['case_num'] == case_num]
    yake_case = yake[yake['case_num'] == case_num]
    print(case_num)
    for pat_note in pn_case.iterrows():
        for pred in yake_case.iterrows():
            for word in pred[1][4].split():
                loc = pat_note[1][2].find(word)
                if loc != -1: locations.append([case_num, pat_note[1][0], pred[1][1], loc, loc+len(word), word])
```

```python
#print('case_num, note_num, feature_num, start, end, word')
def save_submit_file(locations, fp='submit.csv'):
    with open(fp, 'w') as f:
        f.write('id,location')
        prev = locations[0]
        f.write(f"\n{prev[0]}{'{:04d}'.format(prev[1])}_{'{:03d}'.format(prev[2])},{prev[3]} {prev[4]}")
        for i in locations:
            if prev[0:3] == i[0:3]:
                f.write(f";{i[3]} {i[4]}")
            else:
                f.write(f"\n{i[0]}{'{:04d}'.format(i[1])}_{'{:03d}'.format(i[2])},{i[3]} {i[4]}")
            prev = i
```

```python
save_submit_file(locations, 'submit.csv')
```
