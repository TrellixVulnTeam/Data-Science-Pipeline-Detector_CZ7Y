# Usage

```python
from autoviz.AutoViz_Class import AutoViz_Class

AV = AutoViz_Class()
```

## 1.Train Part

```python
filename = "/kaggle/input/ventilator-pressure-prediction/train.csv"
sep = ","
dft = AV.AutoViz(
    filename,
    sep=",",
    depVar="",
    dfte=None,
    header=0,
    verbose=0,
    lowess=False,
    chart_format="svg"
)
```

## 2. Test Part

```python
filename = "/kaggle/input/ventilator-pressure-prediction/test.csv"
sep = ","
dft = AV.AutoViz(
    filename,
    sep=",",
    depVar="",
    dfte=None,
    header=0,
    verbose=0,
    lowess=False,
    chart_format="svg"
)
```
