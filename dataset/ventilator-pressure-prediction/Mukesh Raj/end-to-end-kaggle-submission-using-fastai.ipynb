## Ventilator pressure prediction using fastai and other libraries

```python
#importing the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from fastai.tabular.all import *
%matplotlib inline
```

```python
#loading the training datasets
train_df = pd.read_csv("../input/ventilator-pressure-prediction/train.csv")
train_df.head()
```

```python
#loading the test data set
test_df = pd.read_csv("../input/ventilator-pressure-prediction/test.csv")
test_df.head()
```

```python
#looking into the submission format
sample_sub = pd.read_csv("../input/ventilator-pressure-prediction/sample_submission.csv", index_col=0)
sample_sub.head()
```

```python
splits = RandomSplitter(valid_pct =0.2,seed = 42)(range_of(train_df))
#splits = RandomSplitter(valid_pct=0.2)(range_of(df))
```

```python
cont_names = [col for col in train_df if train_df[col].dtype!="O"]
cont_names
```

```python
cont_names.remove("pressure")
```

```python
cont_names
```

```python
to = TabularPandas(train_df,
                   procs = [Normalize],
                   cont_names = cont_names,
                   y_names = "pressure",
                   splits = splits)
```

```python
to.xs.iloc[:5]
```

```python
dls = to.dataloaders(bs=2048)
```

```python
learn = tabular_learner(dls, metrics=mae)
```

```python
learn
```

```python
dls.valid.show_batch()
```

```python
learn.fit_one_cycle(1)
```

```python
learn.show_results()
```

```python
test_df
```

```python
dl = learn.dls.test_dl(test_df)
```

```python
dl
```

```python
pred = learn.get_preds(dl=dl)
```

```python
pred
```

```python
preds = learn.get_preds(dl=dl)[0].argmax(1).numpy()
preds[:5]
```

```python
preds
```

```python
sample_sub
```

```python
sub = pd.DataFrame({'PassengerId':test_df.id, 'pressure': preds})
sub.to_csv('submission.csv', index=False)
sub.head()
```
