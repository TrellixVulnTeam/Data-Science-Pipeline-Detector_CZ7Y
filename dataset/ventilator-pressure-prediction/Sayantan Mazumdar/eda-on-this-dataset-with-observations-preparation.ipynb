# Processing


```python
def datatorecord(df):
    dic=dict()
    for k in set(df['breath_id']): 
        dic[k]=[]
    for row in tqdm(np.array(df)):
        row=np.array(row)
        try:
            dic[row[1]].append(row[[2,3,5,6,7]])
        except:
            dic[row[1]].append(row[[2,3,5,6]])
    return dic
```

```python
trainrecord=datatorecord(train)
testrecord=datatorecord(test)
```
