# K-means

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
kmeans = KMeans(3)
kmeans.fit(X)
```

```python
identified_clusters = kmeans.predict(X)
identified_clusters[0:6]
```

```python
np.save('clusters.npy',identified_clusters)
```

```python
identified_clusters = kmeans.predict(xtest)
identified_clusters[0:6]
```

```python
np.save('clusters_test.npy',identified_clusters)
```

```python
import pickle
pickle.dump(kmeans,open('kmeans.pkl','wb'))
```
