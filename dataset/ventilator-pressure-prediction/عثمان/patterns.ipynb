```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('../input/ventilator-pressure-prediction/train.csv')
```

```python
start = 1600
num_steps = 10

for R in train.R.unique():
    for C in train.C.unique():
        look = train[
            (train.R==R) &
            (train.C==C)
        ]
        plt.figure(figsize=(25,4))
        plt.title(f'{R}-{C}')
        plt.plot(
            look.time_step[start:start+80*num_steps].cumsum().values,
            look.u_in[start:start+80*num_steps].values,
            label='uin'
        )
        plt.plot(
            look.time_step[start:start+80*num_steps].cumsum().values,
            look.pressure[start:start+80*num_steps].values,
            label='pressure'
        )
        plt.legend()
        plt.show()
```
