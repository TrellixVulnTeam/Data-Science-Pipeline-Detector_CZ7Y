```python
train = pd.read_csv('../input/ventilator-pressure-prediction/train.csv')
test = pd.read_csv('../input/ventilator-pressure-prediction/test.csv')

train = train.drop(['id', 'breath_id'], axis=1)
```

# <p style="background-color:#615154;font-family:newtimeroman;color:#FFFFFF;font-size:100%;text-align:center;border-radius:40px 40px;padding: 3px 0px 3px 0px">Check relation between feature and target variable</p>

```python
# Plot continuous subplot
plt.figure(2, figsize=(10, 40))
plt.subplot(5,1,1)
plt.title('time_step vs pressure')
plt.scatter(train['time_step'], train['pressure'], s=2, color = 'red')

plt.subplot(5,1,2)
plt.title('u_in vs pressure')
plt.scatter(train['u_in'], train['pressure'], s=2)

plt.subplot(5,1,3)
plt.title('u_out vs pressure')
plt.scatter(train['u_out'], train['pressure'], s=50, color = 'green')

plt.subplot(5,1,4)
plt.title('R vs pressure')
plt.scatter(train['R'], train['pressure'], s=50, color = 'red')

plt.subplot(5,1,5)
plt.title('C vs pressure')
plt.scatter(train['C'], train['pressure'], s=50, color = 'blue')
```

```python
plt.figure(3, figsize=(10, 40))
plt.subplot(6,1,1)
plt.title('u_in vs u_out')
plt.scatter(train['u_in'], train['u_out'], s=50)

plt.subplot(6,1,2)
plt.title('u_in vs R')
plt.scatter(train['u_in'], train['R'], s=50, color = 'pink')

plt.subplot(6,1,3)
plt.title('u_in vs C')
plt.scatter(train['u_in'], train['C'], s=50, color = 'green')

plt.subplot(6,1,4)
plt.title('u_out vs R')
plt.scatter(train['u_out'], train['R'], s=50, color = 'red')

plt.subplot(6,1,5)
plt.title('u_out vs C')
plt.scatter(train['u_out'], train['C'], s=50)

plt.subplot(6,1,6)
plt.title('R vs C')
plt.scatter(train['R'], train['C'], s=50, color = 'magenta')
```
