---


**XLA** (Accelerated Linear Algebra) is a domain-specific compiler for linear algebra that can accelerate TensorFlow models with potentially no source code changes. **The results are improvements in speed and memory usage**.

<br>

When a TensorFlow program is run, all of the operations are executed individually by the TensorFlow executor. Each TensorFlow operation has a precompiled GPU/TPU kernel implementation that the executor dispatches to.

XLA provides us with an alternative mode of running models: it compiles the TensorFlow graph into a sequence of computation kernels generated specifically for the given model. Because these kernels are unique to the model, they can exploit model-specific information for optimization.<br><br>

<div class="alert alert-block alert-danger" style="margin: 2em; line-height: 1.7em; font-family: Verdana;">
    <b style="font-size: 16px;">ðŸ›‘ &nbsp; WARNING:</b><br><br>- XLA can not currently compile functions where dimensions are not inferrable: that is, if it's not possible to infer the dimensions of all tensors without running the entire computation<br>
</div>

````python
def flatten_l_o_l(nested_list):
    """ Flatten a list of lists """
    return [item for sublist in nested_list for item in sublist]

def add_features(df, 
                 U_IN_N_FORWARD=6, 
                 U_IN_N_BACKWARD=6, 
                 U_OUT_N_FORWARD=2, 
                 U_OUT_N_BACKWARD=2, 
                 use_rc=True):
    """ TBD """
    
    # Features
    print("\n... Add general features ...\n")
    df['uin_auc'] = df['time_step'] * df['u_in']
    df['uin_auc'] = df.groupby('breath_id')['uin_auc'].cumsum()
    df['uin_csum'] = (df['u_in']).groupby(df['breath_id']).cumsum()
    df['breath_id__u_in__max'] = df.groupby(['breath_id'])['u_in'].transform('max')
    df['breath_id__u_in__diffmax'] = df.groupby(['breath_id'])['u_in'].transform('max') - df['u_in']
    df['breath_id__u_in__diffmean'] = df.groupby(['breath_id'])['u_in'].transform('mean') - df['u_in']
    df['breath_id__u_in__diffmax'] = df.groupby(['breath_id'])['u_in'].transform('max') - df['u_in']
    df['breath_id__u_in__diffmean'] = df.groupby(['breath_id'])['u_in'].transform('mean') - df['u_in']
    df['cross3']= df['time_step']*df['u_in']
    df['cross3_sqd_1']= df['time_step']*df['u_in']**2
    df['cross3_sqd_2']= df['time_step']**2*df['u_in']
    df['cross3_cubed_1']= df['time_step']*df['u_in']**3
    df['cross3_cubed_2']= df['time_step']**3*df['u_in']

    
    # Uin Lag
    print("\t... Add lag and advance UIN features ...")
    for i in range(1, U_IN_N_BACKWARD+1):
        df[f'u_in_{i}_back'] = df.groupby("breath_id")['u_in'].shift(i).fillna(0)
    for i in range(1, U_IN_N_FORWARD+1):
        df[f'u_in_{i}_forw'] = df.groupby("breath_id")['u_in'].shift(-i).fillna(0)
    
    print("\t... Add lag and advance UOUT features ...")
    for i in range(1, U_OUT_N_BACKWARD+1):
        df[f'u_out_{i}_back'] = df.groupby("breath_id")['u_out'].shift(i).fillna(0)
    for i in range(1, U_OUT_N_FORWARD+1):
        df[f'u_out_{i}_forw'] = df.groupby("breath_id")['u_out'].shift(-i).fillna(0)
    
    print("\t... Add UIN and UOUT `diff` features ...")
    for i in range(1, U_IN_N_BACKWARD+1):
        df[f'u_in_diff_{i}_back'] = df['u_in'] - df[f'u_in_{i}_back']
    for i in range(1, U_OUT_N_BACKWARD+1):
        df[f'u_out_diff_{i}_back'] = df['u_out'] - df[f'u_out_{i}_back']
    for i in range(1, U_IN_N_FORWARD+1):
        df[f'u_in_diff_{i}_back'] = df['u_in'] - df[f'u_in_{i}_forw']
    for i in range(1, U_OUT_N_FORWARD+1):
        df[f'u_out_diff_{i}_forw'] = df['u_out'] - df[f'u_out_{i}_forw']
    
    print("\t... Add categorical features ...")
    if use_rc:
        df['R_C'] = df['R'].astype(str)+"_"+df['C'].astype(str)
        df['R'] = df['R']/50
        df['C'] = df['C']/50
        df = pd.get_dummies(df,)
    
    print("\t... Reset dtypes for lower memory usage ...")
    for c in df.columns:
        if c=="u_out":
            df[c] = df[c].astype("uint8")
        elif df[c].dtype=="float64":
            df[c] = df[c].astype("float32")
    
    # Final cleanup
    gc.collect(); gc.collect();
    
    return df
````

````python
def get_lstm_model(print_summary=True, pen_activation="selu"):
    _inputs = tf.keras.layers.Input(shape=(ROWS_PER_BREATH, N_FEATURES))
    x = tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(768, return_sequences=True))(_inputs)
    x = tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(384, return_sequences=True))(x)
    x = tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(256, return_sequences=True))(x)
    x = tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(128, return_sequences=True))(x)
    
    if pen_activation=="linear":
        x = tf.keras.layers.Dense(64, activation="linear", name="head_layer")(x)
    elif pen_activation=="leaky_relu":
        x = tf.keras.layers.Dense(64, activation=None, name="head_layer")(x)
        x = tf.keras.layers.LeakyReLU(alpha=0.02)(x)
    elif pen_activation=="selu":
        x = tf.keras.layers.Dense(64, activation="selu", name="head_layer")(x)
    else:
        raise ValueError()
        
    _outputs = tf.keras.layers.Dense(1, name="classification_layer")(x)
    _model = tf.keras.Model(inputs=_inputs, outputs=_outputs)
    if print_summary: print(_model.summary())
    return _model

class MaskedValidationMAE(tf.keras.callbacks.Callback):
    """ CUSTOM CALLBACK TO PRINT REAL MAE USING CUSTOM VAL DATASET"""
    def on_epoch_end(self, batch, logs={}):
        print(f"\t--> MASKED MAE ON VAL={self.model.evaluate(cb_val_ds, verbose=0)[-1]}\n")
````

````python
USE_SAMPLE_WTS = True
INHALE_WT = 1.00
EXHALE_WT = 0.20

histories = []
all_preds = []

# For demo purposes
for i, (train_indices, val_indices) in tqdm(enumerate(GroupKFold(n_splits=VAL_SPLIT_NTH).
                                                      split(X=train_df[FEATURE_NAMES], 
                                                            y=train_df[LABEL_NAMES], 
                                                            groups=train_df[GROUPBY_NAMES])), 
                                            total=VAL_SPLIT_NTH):
    
    print(f"\n\n\n\n\n...  WORKING ON FOLD {i+1} of {VAL_SPLIT_NTH}  ...\n\n\n")
    
    sub_val_df = train_df.iloc[val_indices].reset_index(drop=True)
    sub_train_df = train_df.iloc[train_indices].reset_index(drop=True)
    
    # Get training data
    sub_train_x_np = RS.transform(sub_train_df[FEATURE_NAMES].to_numpy())
    sub_train_y_np = sub_train_df[LABEL_NAMES].to_numpy()

    # Get validation data
    sub_val_x_np = RS.transform(sub_val_df[FEATURE_NAMES].to_numpy())
    sub_val_y_np = sub_val_df[LABEL_NAMES].to_numpy()
    
    if USE_SAMPLE_WTS:
        sub_train_w_np = sub_train_df["u_out"].apply(lambda x: INHALE_WT if x==0 else EXHALE_WT).to_numpy()
        sub_val_w_np = sub_val_df["u_out"].apply(lambda x: INHALE_WT if x==0 else EXHALE_WT).to_numpy()
    sub_cb_val_w_np = sub_val_df["u_out"].apply(lambda x: 1.0 if x==0 else 0.0).to_numpy()
    
    if USE_SAMPLE_WTS:
        train_ds = tf.data.Dataset.zip((tf.data.Dataset.from_tensor_slices(sub_train_x_np), tf.data.Dataset.from_tensor_slices(sub_train_y_np), tf.data.Dataset.from_tensor_slices(sub_train_w_np)))
        val_ds = tf.data.Dataset.zip((tf.data.Dataset.from_tensor_slices(sub_val_x_np), tf.data.Dataset.from_tensor_slices(sub_val_y_np), tf.data.Dataset.from_tensor_slices(sub_val_w_np)))
    else:
        train_ds = tf.data.Dataset.zip((tf.data.Dataset.from_tensor_slices(sub_train_x_np), tf.data.Dataset.from_tensor_slices(sub_train_y_np)))
        val_ds = tf.data.Dataset.zip((tf.data.Dataset.from_tensor_slices(sub_val_x_np), tf.data.Dataset.from_tensor_slices(sub_val_y_np)))
    cb_val_ds = tf.data.Dataset.zip((tf.data.Dataset.from_tensor_slices(sub_val_x_np), tf.data.Dataset.from_tensor_slices(sub_val_y_np), tf.data.Dataset.from_tensor_slices(sub_cb_val_w_np)))
    
    
    train_ds = train_ds.batch(ROWS_PER_BREATH, drop_remainder=True).cache().shuffle(sub_train_x_np.shape[0]//80).batch(OVERALL_BATCH_SIZE, drop_remainder=True).prefetch(tf.data.AUTOTUNE)
    val_ds = val_ds.batch(ROWS_PER_BREATH, drop_remainder=True).cache().shuffle(sub_val_x_np.shape[0]//80).batch(OVERALL_BATCH_SIZE, drop_remainder=True).prefetch(tf.data.AUTOTUNE)
    cb_val_ds = cb_val_ds.batch(ROWS_PER_BREATH, drop_remainder=True).cache().shuffle(sub_val_x_np.shape[0]//80).batch(OVERALL_BATCH_SIZE, drop_remainder=True).prefetch(tf.data.AUTOTUNE)

    with strategy.scope():
        cb_list = [
            tf.keras.callbacks.EarlyStopping(monitor="val_loss", patience=25, verbose=1, mode="min", restore_best_weights=True),
            tf.keras.callbacks.ReduceLROnPlateau(monitor="val_loss", factor=0.9333, patience=5, verbose=1),
            MaskedValidationMAE(),
            # tf.keras.callbacks.ModelCheckpoint("/kaggle/working/ventnet__{epoch:02d}__loss_{val_loss:.4f}", verbose=1, save_best_only=True, options=save_locally)
        ]
        model = get_lstm_model()

        # MAKE SURE WHEN YOU COMPILE, YOU USE weighted_metrics INSTEAD OF metrics
        model.compile(optimizer=tf.keras.optimizers.Adam(0.0012), 
                      loss="mae", 
                      weighted_metrics=["mae",])

    print(f"\n\n\n... STARTING TRAINING ...\n\n\n")
    histories.append(model.fit(train_ds, validation_data=val_ds, epochs=N_EPOCHS, callbacks=cb_list))
    model.save(f"./best_model_fold_{i}", options=save_locally)
    all_preds.append(model.predict(test_ds).reshape(-1)[:N_TEST])
````
