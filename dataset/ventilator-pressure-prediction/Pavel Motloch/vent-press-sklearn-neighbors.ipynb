# Introduction

A short code using the sklearn nearest neighbor routine. For each test breath, find the train breath with the most similar u_in (MAE metric) and use its pressure.

# Import libraries and data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import NearestNeighbors
```

```python
train = pd.read_csv('../input/ventilator-pressure-prediction/train.csv')
test  = pd.read_csv('../input/ventilator-pressure-prediction/test.csv')
sub   = pd.read_csv('../input/ventilator-pressure-prediction/sample_submission.csv', index_col = "id")
```

```python
BL = 80 #Breath length - how many data points we have for each breath
```

# Extract u_in as a numpy array, check

```python
samples = train['u_in'].values.reshape((-1, BL))
test_samples = test['u_in'].values.reshape((-1, BL))
```

```python
plt.plot(samples[0]);
```

# Define the estimator and fit it to the train data

```python
neigh = NearestNeighbors(n_neighbors=1, metric = 'manhattan')
neigh.fit(samples)
```

# For each test breath, find the nearest neighbor in the train sample and use its pressure

```python
train_p_loc = train.columns.get_loc('pressure') # integer location of pressure
sub_p_loc  = sub.columns.get_loc('pressure')

for idx in range(len(test_samples)):
    nn = neigh.kneighbors([test_samples[idx]], 1, return_distance=False)[0,0]
    sub.iloc[idx*BL:(idx+1)*BL, sub_p_loc] = train.iloc[nn*BL:(nn+1)*BL, train_p_loc].values
```
