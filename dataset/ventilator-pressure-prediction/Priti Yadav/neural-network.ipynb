```python
import pandas as pd
import numpy as np

# Make numpy values easier to read.
np.set_printoptions(precision=3, suppress=True)

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.layers.experimental import preprocessing

import gc
```

```python
# # load the dataset
# train = loadtxt('../input/ventilator-pressure-prediction/train.csv', delimiter=',')
# test = loadtxt('../input/ventilator-pressure-prediction/test.csv', delimiter=',')
train = pd.read_csv('../input/ventilator-pressure-prediction/train.csv')
test = pd.read_csv('../input/ventilator-pressure-prediction/test.csv')
sample = pd.read_csv('../input/ventilator-pressure-prediction/sample_submission.csv')
```

```python
train.head()
```

```python
X = train.copy()
y = X.pop('pressure')
```

```python
X = np.array(X)
X
```

```python
model = tf.keras.Sequential([
  layers.Dense(64),
  layers.Dense(32),
  layers.Dense(1)
])

model.compile(loss = tf.losses.MeanSquaredError(),
                      optimizer = tf.optimizers.Adam())
```

```python
# from keras import callbacks
earlystopping = tf.keras.callbacks.EarlyStopping(monitor ="val_loss", 
                                        mode ="min", patience = 5, 
                                        restore_best_weights = True)
```

```python
model.fit(X, y, batch_size = 32, 
                    epochs = 15, 
                    callbacks =[earlystopping])
```

```python
gc.collect()
```

```python
# make class predictions with the model
predictions = model.predict_classes(X)
# summarize the first 5 cases
for i in range(5):
	print('%s => %d (expected %d)' % (X[i].tolist(), predictions[i], y[i]))
```

```python
gc.collect()
```

```python
predicted = model.predict(test)
predicted
```

```python
test_result = test
test_result['pressure'] = predicted
```

```python
submit_result = test_result[['id','pressure']]
submit_result
```

```python
submit_result.to_csv('submission.csv', index=False)
```
