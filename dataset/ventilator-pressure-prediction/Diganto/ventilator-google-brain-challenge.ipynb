```python
import pandas as pd
import numpy as nm

train= pd.read_csv('../input/ventilator-pressure-prediction/train.csv')
train.info()
train.describe()
```

```python

#feature engineering
train["u_in_cumsum"] = (train['u_in']).groupby(train['breath_id']).cumsum()
train["time_step_cumsum"] = (train['time_step']).groupby(train['breath_id']).cumsum()
train["area"] = train['R']*train['C']
#time lag variable
train["time_diff"] = (train['time_step']).groupby(train['breath_id']).diff(1)
train["time_diff2"] = (train['time_step']).groupby(train['breath_id']).diff(2)
train["time_diff3"] = (train['time_step']).groupby(train['breath_id']).diff(3)
train["time_diff4"] = (train['time_step']).groupby(train['breath_id']).diff(4)
train["time_diff5"] = (train['time_step']).groupby(train['breath_id']).diff(5)
train["time_diff6"] = (train['time_step']).groupby(train['breath_id']).diff(6)
train["time_diff7"] = (train['time_step']).groupby(train['breath_id']).diff(7)
train["time_diff8"] = (train['time_step']).groupby(train['breath_id']).diff(8)

train['u_in_lag1'] = train.groupby('breath_id')['u_in'].shift(1)
train['u_in_lag2'] = train.groupby('breath_id')['u_in'].shift(2)
train['u_in_lag3'] = train.groupby('breath_id')['u_in'].shift(3)
train['u_in_lag4'] = train.groupby('breath_id')['u_in'].shift(4)
train['u_in_lag5'] = train.groupby('breath_id')['u_in'].shift(5)
train['u_in_lag6'] = train.groupby('breath_id')['u_in'].shift(6)
train['u_in_lag7'] = train.groupby('breath_id')['u_in'].shift(7)
train['u_in_lag8'] = train.groupby('breath_id')['u_in'].shift(8)
train['u_in_lag9'] = train.groupby('breath_id')['u_in'].shift(9)
train['u_in_lag10'] = train.groupby('breath_id')['u_in'].shift(10)
train['u_in_lag11'] = train.groupby('breath_id')['u_in'].shift(11)
train['u_in_lag12'] = train.groupby('breath_id')['u_in'].shift(12)
train['u_in_lag13'] = train.groupby('breath_id')['u_in'].shift(13)
train['u_in_lag14'] = train.groupby('breath_id')['u_in'].shift(14)
train['u_in_lag15'] = train.groupby('breath_id')['u_in'].shift(15)
train['u_in_lag16'] = train.groupby('breath_id')['u_in'].shift(16)
train['u_in_lag17'] = train.groupby('breath_id')['u_in'].shift(17)
train['u_in_lag18'] = train.groupby('breath_id')['u_in'].shift(18)
train['u_in_lag19'] = train.groupby('breath_id')['u_in'].shift(19)
train['u_in_lag20'] = train.groupby('breath_id')['u_in'].shift(20)
train['u_in_lag21'] = train.groupby('breath_id')['u_in'].shift(21)
train['diff'] = train['u_in_lag1']-train['u_in_lag2'] 
train['diff2'] = train['u_in_lag1']-train['u_in_lag3'] 


#Missign value treatment
train = train.fillna(0)
train.head(10)
```

```python
#Creating Train Data
y = train['pressure']
xtrain = train.drop('pressure',axis = 1)
xtrain = xtrain.drop('id',axis = 1)
xtrain.head(10)
```

```python
#XG Boost model with GPU Hist method
from xgboost import XGBRegressor
xgb=XGBRegressor(learning_rate = 0.1, n_estimators = 10000,max_depth =9,
                n_jobs = 16, tree_method = 'gpu_hist')
model_xgb = xgb.fit(xtrain,y)


```

```python
#Reading the test data
test= pd.read_csv('../input/ventilator-pressure-prediction/test.csv')
xtest = test.drop('id',axis = 1)

#Feature Changes
xtest['u_in_cumsum'] = (xtest['u_in']).groupby(xtest['breath_id']).cumsum()
xtest['time_step_cumsum'] = (xtest['time_step']).groupby(xtest['breath_id']).cumsum()
xtest['area'] = xtest['R']*xtest['C']
xtest['time_diff'] = (xtest['time_step']).groupby(xtest['breath_id']).diff(1)
xtest['time_diff2'] = (xtest['time_step']).groupby(xtest['breath_id']).diff(2)
xtest['time_diff3'] = (xtest['time_step']).groupby(xtest['breath_id']).diff(3)
xtest['time_diff4'] = (xtest['time_step']).groupby(xtest['breath_id']).diff(4)
xtest['time_diff5'] = (xtest['time_step']).groupby(xtest['breath_id']).diff(5)
xtest['time_diff6'] = (xtest['time_step']).groupby(xtest['breath_id']).diff(6)
xtest['time_diff7'] = (xtest['time_step']).groupby(xtest['breath_id']).diff(7)
xtest['time_diff8'] = (xtest['time_step']).groupby(xtest['breath_id']).diff(8)
xtest['u_in_lag1'] = xtest.groupby('breath_id')['u_in'].shift(1)
xtest['u_in_lag2'] = xtest.groupby('breath_id')['u_in'].shift(2)
xtest['u_in_lag3'] = xtest.groupby('breath_id')['u_in'].shift(3)
xtest['u_in_lag4'] = xtest.groupby('breath_id')['u_in'].shift(4)
xtest['u_in_lag5'] = xtest.groupby('breath_id')['u_in'].shift(5)
xtest['u_in_lag6'] = xtest.groupby('breath_id')['u_in'].shift(6)
xtest['u_in_lag7'] = xtest.groupby('breath_id')['u_in'].shift(7)
xtest['u_in_lag8'] = xtest.groupby('breath_id')['u_in'].shift(8)
xtest['u_in_lag9'] = xtest.groupby('breath_id')['u_in'].shift(9)
xtest['u_in_lag10'] = xtest.groupby('breath_id')['u_in'].shift(10)
xtest['u_in_lag11'] = xtest.groupby('breath_id')['u_in'].shift(11)
xtest['u_in_lag12'] = xtest.groupby('breath_id')['u_in'].shift(12)
xtest['u_in_lag13'] = xtest.groupby('breath_id')['u_in'].shift(13)
xtest['u_in_lag14'] = xtest.groupby('breath_id')['u_in'].shift(14)
xtest['u_in_lag15'] = xtest.groupby('breath_id')['u_in'].shift(15)
xtest['u_in_lag16'] = xtest.groupby('breath_id')['u_in'].shift(16)
xtest['u_in_lag17'] = xtest.groupby('breath_id')['u_in'].shift(17)
xtest['u_in_lag18'] = xtest.groupby('breath_id')['u_in'].shift(18)
xtest['u_in_lag19'] = xtest.groupby('breath_id')['u_in'].shift(19)
xtest['u_in_lag20'] = xtest.groupby('breath_id')['u_in'].shift(20)
xtest['u_in_lag21'] = xtest.groupby('breath_id')['u_in'].shift(21)
xtest['diff'] = xtest['u_in_lag1']-xtest['u_in_lag2'] 
xtest['diff2'] = xtest['u_in_lag1']-xtest['u_in_lag3'] 

#Fill missing values
xtest = xtest.fillna(0)
submission_xgb=model_xgb.predict(xtest)
```

```python
#Submission
sub2 = pd.DataFrame(submission_xgb)
df_submit = pd.concat([test.id, sub2], axis=1)
df_submit.columns = ["id","pressure"]
df_submit.to_csv('./submission.csv',index = False)
```
