# Baseline or Possibly easy to add new feature

I entered this challenge quite late i'm doing experiments with baselines proposed by users and i came up with this new "baseline" , it doesn't have great perfomance but it could be used as feature for other model.
Beware of overfitting, maybe some kind of holdout set could be used to avoid leakage of validation information into the new feature.
If someone already posted something similar i'm sorry i didn't find it.

```python
import numpy as np 
import pandas as pd 
path = "../input/ventilator-pressure-prediction/"
train = pd.read_csv(path + 'train.csv',usecols=["id","pressure"])
test  = pd.read_csv(path + 'test.csv',usecols=["id"])

#samples are ordered in group of 80 in time order (should check) in case reorder
train["step"]=(train["id"]-1)%80
test["step"]=(train["id"]-1)%80
#get pressure grouped by their step 
stepWiseTarget=(train['pressure']).groupby(train['step'])
#Compute mean and median
mean=stepWiseTarget.mean().to_numpy()
median=stepWiseTarget.median().to_numpy()

choice="mean"

```

```python
if choice=="mean":
    test["pressure"]=test["step"].apply(lambda x: mean[x])
    #Score is together with this notebook , still experimenting
else:
    test['pressure']=test["step"].apply(lambda x: median[x])
    #public score 6.2442 but it increases performance if used as feature in other models
    
test.drop("step",axis=1,inplace=True)

test.to_csv("submission.csv",index=False)
```
