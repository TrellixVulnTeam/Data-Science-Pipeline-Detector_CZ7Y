## Project Explained

**Project Scope** - Clinitians use mechanical ventilators to give oxygen to a sedated patient, But operating a mechanical ventilator is a labourous procedure.
There types of ventilators who can work based on the lung output are extremely costly. So there is a gap for a lung ventilation simulator based on deep learning which can adapt to each type of patient. This would reduce the cost of ventilators a lot

**Parameters**
* U_in - The amount of air allowed into the lungs during inhalation process ( or expansion )
* U_out - The amount of air pumped out of lungs during exhalation ( or contraction )
* Pressure - The mesure of pressure when U_In and U_out is happening
* C - Complience of the lungs or how easily can we strech the lungs. 3 values given ( 10, 20, 50)
* R - Resistance of lungs. How easy can air go in and out of lungs. 3 values (5, 20, 50)

**Training Data**
* 7 Columns including ID
* 2 data types int64 & float 64

## Finding the unique values amoung the variables

Here we need to find the combinations of ID ,R and C to plot them against the time stamp.
So lets observe the unique values among the following variables
* ID
* R
* C

```python
print("Number of unique breath ID's : ", train['breath_id'].unique() )
print('\n')
print("Number of unique R : ", train['R'].unique() )
print('\n')
print("Number of unique C : ", train['C'].unique() )
print('\n')
```

## Lets see the occurance of R & C with each of the breath ID

```python
# Read all ID's into a variable
Id = train['breath_id']
# Enumerate ID and see the unique R & C values
for i,Id in enumerate(train['breath_id'].unique()):
    print("ID : ",Id)
    print(train[train['breath_id']==Id]['R'].unique())
    print(train[train['breath_id']==Id]['C'].unique())
    if i==5:
        break
```

**Here we can observe that each breath ID has specific R and C comination**

```python
import matplotlib.pyplot as plt
def plot_graph(ID):
    column = ['u_in', 'u_out', 'pressure']
    breath_df = train[train['breath_id']==ID]
    #Now plot the figure
    plt.figure(figsize=(12,4))
    for col in column:
        plt.plot(breath_df['time_step'], breath_df[col], label=col)
    r,c = breath_df[['R','C']].values[0]
    plt.legend()
    plt.title(f'Breath ID: {ID} R={r} C={c}')
    #print(breath_df)
    
    
#visulalize based on breath ID's
for i,Id in enumerate(train['breath_id'].unique()):
    if i==10:
        break
    plot_graph(Id)
```

Observed points
1. The u-out doest show much variations with time
1. u-in and pressure show some relationship with
1. pressure drop as there is a drop in u-in.
1. u-out has a small step when the u-in has a steep drop.
