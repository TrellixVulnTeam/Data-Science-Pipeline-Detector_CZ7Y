```python
import pandas as pd
import plotly.graph_objects as go
```

```python
train_df = pd.read_csv('../input/ventilator-pressure-prediction/train.csv')
test_df = pd.read_csv('../input/ventilator-pressure-prediction/test.csv')
```

#### Parameter R = [5, 20, 50], C = [10,20,50]

```python
print(train_df['R'].value_counts())
print(train_df['C'].value_counts())
```

## 1. Case of [R=5, C=10] graph plot

```python
fig = go.Figure()
num_id = list(train_df['breath_id'].unique())
sample_num_id = num_id[0:200]

for i in sample_num_id:
    
    idx = train_df.loc[train_df['breath_id'] == i,'breath_id'].unique()[0]
    R = train_df.loc[train_df['breath_id'] == i,'R'].unique()[0]
    C = train_df.loc[train_df['breath_id'] == i,'C'].unique()[0]
    
    if (R == 5) & (C == 10): 
        fig.add_trace(go.Scatter(x=train_df.loc[train_df['breath_id'] == i,'time_step'],
                                 y=train_df.loc[train_df['breath_id'] == i,'pressure'],
                                 name=f'breath_id : {idx}, R : {R}, C : {C}'))

fig.show()
```
