```python
import numpy as np
import pandas as pd

import os, sys
from sklearn.cluster import SpectralClustering
from matplotlib import pyplot as plt

#from sklearn.cluster import KMeans, AgglomerativeClustering
PATH = "../input/ventilator-pressure-prediction"


df = pd.read_csv(f"{PATH}/train.csv")
print(df.shape)
df.head()
```

```python
#plt.scatter(df['pressure'], df['u_in'], alpha=0.7, edgecolors='b')
```

```python
#plt.scatter(df['pressure'], df['C'], alpha=0.7, edgecolors='b')
```

```python
df = df.groupby('breath_id')['u_out','u_in','R','C'].agg(np.mean)
```

```python
sc2 = SpectralClustering(n_clusters=4, random_state=0)
sc_clustering2 = sc2.fit(df[['u_out','u_in']])
plt.scatter(df['u_out'], df['u_in'], c=sc_clustering2.labels_, cmap='rainbow', alpha=0.9)
```

```python
sc3 = SpectralClustering(n_clusters=3, random_state=0)
sc_clustering3 = sc3.fit(df[['R','C']])
plt.scatter(df['R'], df['C'], c=sc_clustering3.labels_, cmap='rainbow', alpha=0.7, edgecolors='b')
```
