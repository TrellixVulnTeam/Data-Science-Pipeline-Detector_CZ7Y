```python

from keras.models import Sequential
from keras.layers import Dense,BatchNormalization,Dropout
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
import tensorflow as tf
```

```python
model = Sequential()
model.add(Dense(16, input_dim=5, kernel_initializer='normal', activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(32, kernel_initializer='normal', activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.2))
model.add(Dense(64, kernel_initializer='normal',activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(128, kernel_initializer='normal', activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.2))
model.add(Dense(256, kernel_initializer='normal',activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(512, kernel_initializer='normal', activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.2))
model.add(Dense(128, kernel_initializer='normal',activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(64, kernel_initializer='normal', activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(8, kernel_initializer='normal', activation='relu'))
model.add(Dense(1, kernel_initializer='normal'))
    # Compile mode
model.compile(optimizer='adam',
              loss='mean_squared_error',
              metrics=['accuracy'])
early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_accuracy', patience=30, restore_best_weights=True)
```

```python
model.summary()
```

```python
Train_data= pd.read_csv("/kaggle/input/ventilator-pressure-prediction/train.csv")
Test_data=pd.read_csv("/kaggle/input/ventilator-pressure-prediction/test.csv")
```

```python
X=Train_data.drop(['breath_id'],axis=1)
X=X.drop(['id'],axis=1)
#X=X.drop( ['R'],axis=1)
#X=X.drop(['C'],axis=1)
#X=X.drop( ['u_in'],axis=1)
X=X.drop( ['pressure'],axis=1)
```

```python
Y=Train_data["pressure"]
```

```python
X_train, X_val, Y_train, Y_val = train_test_split(X, Y, test_size=0.3, random_state=1)

```

```python
X_train.shape
```

```python
Train_data =X_train#np.expand_dims(X_train, axis=-1)
Valid_data=X_val#np.expand_dims(X_val, axis=-1)
Train_abel=Y_train
Valid_label=Y_val
Valid_label= np.array(Valid_label)
Train_abel = np.array(Train_abel)
Valid_label=Valid_label.reshape(-1)
Train_abel=Train_abel.reshape(-1)
```

```python
device_name = tf.test.gpu_device_name()
print(device_name)
```

```python
tf.config.experimental_run_functions_eagerly(True)
history = model.fit(Train_data, Train_abel, epochs=4,batch_size=100000, 
                    validation_data=(Valid_data, Valid_label),callbacks=[early_stopping])
```

```python
Xx=Test_data.drop(['breath_id'],axis=1)
Xx=Xx.drop(['id'],axis=1)
#X=X.drop( ['R'],axis=1)
#X=X.drop(['C'],axis=1)
#X=X.drop( ['u_in'],axis=1)


```

```python
predictions = model.predict(Xx,batch_size=100000)
```
