Configuration of python setting and read csv files.

Pythonの機能読み込みとcsvファイルの読み込み

```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns

df1 = pd.read_csv("../input/ventilator-pressure-prediction/train.csv")
df2 = pd.read_csv("../input/ventilator-pressure-prediction/test.csv")
df3 = pd.read_csv("../input/ventilator-pressure-prediction/sample_submission.csv")
```

find the data package is 80 index from both df1 and df2

全て80個のデータで１セットであることがわかった

convine train.csv with test.csv as named df12

trainとtestデータを合体する

```python
df12=pd.concat([df1,df2],axis=0)
```

sort the value using "breath_id" as first priority and "time_step " as second priority

breath_idを第一優先、time_stepを第二優先としてデータをソートする

```python
# sort df12 using "breath_id" and "time_step" for realizing correct data order.
# If don't make correct order, error happens when data slicing 80 package.
df12sorted=df12.sort_values(["breath_id","time_step"]).reset_index(drop=True)
```

merged data will start to slice as 80 index/one packeage

合体したデータを80行ワンセットで切り出す

```python
# k is number of slicing one package.
k=80
n=len(df12sorted)


dfs = [df12sorted.loc[i:i+k-1, :] for i in range(0, n, k)]
for DF in dfs:
    DF=dfs
```

To make/confirm the current situation of each breath_id with time-history(Not actuall time-history at this moment) waveform graph

データを時系列的に並べて図示する。（この段階ではX軸は時間ではなく、データ個数）

below is an example to show the graph from #0 to #10 of "breath_id"

0~10のグラフを描画してみる。

```python
for i in range(11):
    breath = DF[i].loc[DF[i]['breath_id']==i].reset_index(drop=True)
    print('breath_id:', i)
    breath[['u_in', 'u_out', 'pressure']].plot()
    plt.show()
```
