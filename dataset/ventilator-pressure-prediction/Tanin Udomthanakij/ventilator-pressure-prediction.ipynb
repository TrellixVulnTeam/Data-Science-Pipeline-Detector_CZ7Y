```python
def plot_graph(df, ID):
    cols = ['u_in', 'u_out', 'pressure']
    df_breath = df[df['breath_id']==ID]
    plt.figure(figsize=(12, 4))
    for col in ['u_in', 'u_out', 'pressure']:
        plt.plot(df_breath['time_step'], df_breath[col], label=col)
    r, c = df_breath[['R', 'C']].values[0]
    
    plt.legend()
    plt.title(f'Breath_ID: {ID}  R={r}, C={c}')
```

```python
#visualize each breath id
for i, Id in enumerate(train['breath_id'].unique()):
    if i==6:
        break
    plot_graph(train, Id)
```
