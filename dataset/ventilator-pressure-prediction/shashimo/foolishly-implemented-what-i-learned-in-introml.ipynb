```python
#Import helpful libraries
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split

#Import CSV files
train_data=pd.read_csv("/kaggle/input/ventilator-pressure-prediction/train.csv")
practice_train_data=train_data.sample(n=20000)
test_data=pd.read_csv("/kaggle/input/ventilator-pressure-prediction/test.csv")
train_data.describe()
#train_data.head()
```

```python
#Create X,y
features=['breath_id','R','C','time_step','u_in','u_out']
X=practice_train_data[features]
y=practice_train_data.pressure

#split data to validation
train_X,val_X,train_y,val_y=train_test_split(X,y,random_state=1)
```

```python
#define model, fit, predict and caliculate MAE
rf_model = RandomForestRegressor(random_state=1)
rf_model.fit(train_X, train_y)
rf_val_predictions = rf_model.predict(val_X)
rf_val_mae = mean_absolute_error(rf_val_predictions,val_y)
print(f"Validation MAE for Random Forest Model:{rf_val_mae}")
```

```python
#Test Data Prediction
test_X=test_data[features]
test_preds=rf_model.predict(test_X)
```

```python
#check the results just to be sure
print(test_preds[0:20])
```

```python
output=pd.DataFrame({'id': test_data.id,'pressure': test_preds})
output.to_csv('submission.csv', index=False)
print('end')
```
