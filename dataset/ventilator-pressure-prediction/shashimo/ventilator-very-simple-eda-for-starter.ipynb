```python
# Import helpful libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Importing CSV files
X_train_full=pd.read_csv("/kaggle/input/ventilator-pressure-prediction/train.csv")
X_test_full=pd.read_csv("/kaggle/input/ventilator-pressure-prediction/test.csv")
```

# 1.Simple EDA

- The following histogram shows that R can take 5, 20, 50 and C can take 10, 20, 50. 
- So we know that there are 9 possible combinations of R and C.

```python
#ideas from this notebook: https://www.kaggle.com/currypurin/ventilator-pressure-simple-eda

X_train_full['R_C'] = [f'{r}_{c}' for r, c in zip(X_train_full['R'], X_train_full['C'])]
sns.countplot(data=X_train_full, x='R_C')
```

- Next, in order to understand what R and C mean, I would like to plot u_in and u_out and pressure for several breaths, keeping the combination of R and C fixed.

```python
#ideas from this notebook: https://www.kaggle.com/theoviel/deep-learning-starter-simple-lstm

r=5
c=50
num_of_plots_to_display=5
sample_number=6

def rcplot(num):
    df_checkrc=X_train_full[X_train_full['R']==r]
    df_checkrc=df_checkrc[df_checkrc['C']==c]
    df_checkrc=df_checkrc[df_checkrc['breath_id']==df_checkrc['breath_id'].unique()[num]]
    plt.figure(figsize=(6,2))
    for col in ['u_in','u_out','pressure']:
        plt.plot(df_checkrc['time_step'],df_checkrc[col],label=col)
    plt.legend()
    plt.title(f'Sample {num} - R={r}, C={c}')

for i in range(num_of_plots_to_display):
    rcplot(i+sample_number)
```

# 2.Conclusion


- Try using the above code to create plots with various patterns. Then, compare them and you will get various insights.
- For example, keep R constant and look at the results of varying C .
- A larger C means that the lungs are more easily stretched. So, in the plot with C of 50, we can see that the air can easily escape into the lungs, so the pressure is not likely to be high (note that the scale of the axis is different for each plot).
- Conversely, look at the results of keeping C constant and varying R.A large R means that the airway resistance is large and air cannot flow. So, in the plot with R of 50, we can see that the pressure is higher.
