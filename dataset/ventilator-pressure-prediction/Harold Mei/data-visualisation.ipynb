## Visualize the data

```python
import pandas as pd
import numpy as np

from matplotlib.pyplot import plot

base = '../input/ventilator-pressure-prediction/'
df1 = pd.read_csv(base+'train.csv')
df2 = pd.read_csv(base+'test.csv')
```

## Pivot for the time series data

```python
df1['tid'] = (df1['id'] - 1)%80
df1_u_in = df1.pivot(index='breath_id', columns='tid', values='u_in')
df1_u_out = df1.pivot(index='breath_id', columns='tid', values='u_out')
df1_u_pressure = df1.pivot(index='breath_id', columns='tid', values='pressure')

df2['tid'] = (df2['id'] - 1)%80
df2_u_in = df2.pivot(index='breath_id', columns='tid', values='u_in')
df2_u_out = df2.pivot(index='breath_id', columns='tid', values='u_out')
```

## plot the 'u_in' column

```python
plot(range(80), df1_u_in.loc[1].to_list(), 'b-')
plot(range(80), df1_u_in.loc[2].to_list(), 'r-')
plot(range(80), df1_u_in.loc[3].to_list(), 'y-')
plot(range(80), df1_u_in.loc[4].to_list(), 'g-')
```

## plot the target pressure

```python
plot(range(80), df1_u_pressure.loc[1].to_list(), 'b-')
plot(range(80), df1_u_pressure.loc[2].to_list(), 'r-')
plot(range(80), df1_u_pressure.loc[3].to_list(), 'y-')
plot(range(80), df1_u_pressure.loc[4].to_list(), 'g-')
```

## plot the 'u_out' column

```python
plot(range(80), df1_u_out.loc[1].to_list(), 'b-')
plot(range(80), df1_u_out.loc[2].to_list(), 'r-')
plot(range(80), df1_u_out.loc[3].to_list(), 'y-')
plot(range(80), df1_u_out.loc[4].to_list(), 'g-')
```

```python
df1_u = df1_u_in + df1_u_out
df2_u = df2_u_in + df2_u_out

plot(range(80), df1_u.loc[1].to_list(), 'b-')
plot(range(80), df1_u.loc[2].to_list(), 'r-')
plot(range(80), df1_u.loc[3].to_list(), 'y-')
plot(range(80), df1_u.loc[4].to_list(), 'g-')
```

```python
df1_u['mean'] = df1_u.mean(axis=1)
df1_u['std'] = df1_u.std(axis=1)
df1_u['min'] = df1_u.min(axis=1)
df1_u['max'] = df1_u.max(axis=1)

df2_u['mean'] = df2_u.mean(axis=1)
df2_u['std'] = df2_u.std(axis=1)
df2_u['min'] = df2_u.min(axis=1)
df2_u['max'] = df2_u.max(axis=1)
```
