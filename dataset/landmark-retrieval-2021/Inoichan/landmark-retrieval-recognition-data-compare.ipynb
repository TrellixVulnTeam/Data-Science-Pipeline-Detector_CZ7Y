```python
import pandas as pd
from glob import glob
```

# train

```python
retrieval_train = [i.split('/')[-1] for i in glob('../input/landmark-retrieval-2021/train/*/*/*/*.jpg')]
recognition_train = [i.split('/')[-1] for i in glob('../input/landmark-recognition-2021/train/*/*/*/*.jpg')]

print(len(list(set(retrieval_train) & set(recognition_train))))
print(len(set(retrieval_train)))
print(len(set(recognition_train)))
```

# test

```python
retrieval_test = [i.split('/')[-1] for i in glob('../input/landmark-retrieval-2021/test/*/*/*/*.jpg')]
recognition_test = [i.split('/')[-1] for i in glob('../input/landmark-recognition-2021/test/*/*/*/*.jpg')]

print(len(set(retrieval_test) & set(recognition_test)))
print(len(set(retrieval_test)))
print(len(set(recognition_test)))
```
