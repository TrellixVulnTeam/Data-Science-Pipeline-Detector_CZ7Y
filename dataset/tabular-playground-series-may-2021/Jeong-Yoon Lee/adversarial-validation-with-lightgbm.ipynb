# Adversarial Validation with AutoLGB

## Introduction to Adversarial Validation

Adversarial validation is an approach to measure the difference in feature distributions between the training and test data sets. We train a binary classifier with a new target variable indicating whether a sample belongs to the test data set (1) or not (0). 

If the adversarial validation classifier performs well (e.g. AUC score >> 50%), the feature distributions are different between the training and test data sets. In this case, good performances on the training data set might not guarantee the same on the test data set. 

If the adversarial validation classifier performs poorly (e.g. AUC score ~ 50%), the feature distirubtions are similar between the training and test data sets, and we may not worry much about overfitting to the training data set during model training.

```python
import kaggler
from kaggler.model import AutoLGB
print(kaggler.__version__)
```

```python
plt.style.use('fivethirtyeight')
pd.set_option('max_columns', 100)
simplefilter('ignore')
```

```python
data_dir = Path('../input/tabular-playground-series-may-2021')
train_file = data_dir / 'train.csv'
test_file = data_dir / 'test.csv'
sample_file = data_dir / 'sample_submission.csv'

id_col = 'id'
target_col = 'target'

n_fold = 5
seed = 42
```

```python
trn = pd.read_csv(train_file, index_col=id_col)
tst = pd.read_csv(test_file, index_col=id_col)
sub = pd.read_csv(sample_file, index_col=id_col)
print(trn.shape, tst.shape, sub.shape)
```

```python
n_trn = trn.shape[0]
df = pd.concat([trn.drop(target_col, axis=1), tst], axis=0)
print(df.shape)
```

```python
cv = StratifiedKFold(n_splits=n_fold, shuffle=True, random_state=seed)
X = df
y = pd.Series(np.concatenate([np.zeros(n_trn,), np.ones(df.shape[0] - n_trn,)]))
p = np.zeros_like(y, dtype=float)
for i, (i_trn, i_val) in enumerate(cv.split(X, y)):
    if i == 0:
        clf = AutoLGB(objective='binary', metric='auc', random_state=seed)
        clf.tune(X.iloc[i_trn], y[i_trn])
        features = clf.features
        params = clf.params
        n_best = clf.n_best
        print(f'{n_best}')
        print(f'{params}')
        print(f'{features}')
    
    trn_data = lgb.Dataset(X.iloc[i_trn], y[i_trn])
    val_data = lgb.Dataset(X.iloc[i_val], y[i_val])
    clf = lgb.train(params, trn_data, n_best, val_data, verbose_eval=100)
    p[i_val] = clf.predict(X.iloc[i_val])
    print(f'CV #{i + 1} AUC: {roc_auc_score(y[i_val], p[i_val]):.6f}')
```

```python
print(f'CV AUC: {roc_auc_score(y, p):.6f}')
```

## Conclusion

Adversarial validation AUC score is close to 50%. Therefore, we can say that the training and test data sets are similar in terms of feature distributions. In other words, no big shake-up expected at the end of the competition. :)
