# CatBoost

The most basic approach to generating predictions, using default settings of catboost.

1) load data

2) train and predict

3) save output submission

## Load Data

```python
import pandas as pd
train = pd.read_csv("../input/tabular-playground-series-may-2021/train.csv")
test = pd.read_csv("../input/tabular-playground-series-may-2021/test.csv")
y_train = train['target']
x_train = train[[coln for coln in train.columns if "feature" in coln]]
x_test = test[[coln for coln in train.columns if "feature" in coln]]
```

## Train and Predict

Generate a predictions in three lines of code using catboost.


```python
from catboost import CatBoostClassifier

model = CatBoostClassifier(loss_function='MultiClassOneVsAll', 
                           random_seed = 0)
model.fit(x_train, y_train, verbose = 1000)
preds = model.predict_proba(x_test)
```

## Save Outputs

```python
sub = pd.read_csv("../input/tabular-playground-series-may-2021/sample_submission.csv")
for idx, class_name in enumerate(model.classes_):
    sub[class_name]=preds[:,idx]
sub.to_csv("CatBoost.csv",index=False)
```
