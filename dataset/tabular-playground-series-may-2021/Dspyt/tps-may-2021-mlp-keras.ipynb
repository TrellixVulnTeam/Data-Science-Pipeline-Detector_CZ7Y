```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

from keras.models import Sequential
from keras.layers import Dense, Dropout, BatchNormalization
from keras.utils.np_utils import to_categorical
from sklearn.preprocessing import LabelEncoder, StandardScaler, MinMaxScaler
from keras.utils import np_utils
from keras.optimizers import RMSprop, SGD
```

```python
sub  = pd.read_csv("/kaggle/input/tabular-playground-series-may-2021/sample_submission.csv")
train= pd.read_csv("/kaggle/input/tabular-playground-series-may-2021/train.csv")
test = pd.read_csv("/kaggle/input/tabular-playground-series-may-2021/test.csv")
```

```python
X_train, y_train = train.iloc[:,1:51], train.iloc[:,[51]]
X_test = test.iloc[:,1:51]
```

```python
sc=MinMaxScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc  = sc.fit_transform(X_test)
```

```python
def one_hot_encode(arr):
    uniques, ids = np.unique(arr, return_inverse=True)
    return np_utils.to_categorical(ids, len(uniques))

y_ohe = one_hot_encode(y_train)
```

```python
model = Sequential([
Dense(512,input_shape=(50,), activation='relu'),
BatchNormalization(),
Dropout(0.3),
Dense(256,activation='relu'),
BatchNormalization(),
Dropout(0.2),
Dense(128,activation='relu'),
BatchNormalization(),
Dropout(0.2),
Dense(4, activation='softmax'),
])
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X_train_sc, y_ohe, batch_size=10, epochs=100)
```

```python
model.summary()
```

```python
output = model.predict(X_test_sc)
```

```python
submit = pd.DataFrame(output, columns=['Class_1','Class_2','Class_3','Class_4'])
submit['id'] = sub.id
submit.set_index('id').to_csv('export.csv',index=True)
```
