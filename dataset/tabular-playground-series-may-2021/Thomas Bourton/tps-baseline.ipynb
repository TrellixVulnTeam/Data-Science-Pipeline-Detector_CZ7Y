```python
import numpy as np
import pandas as pd
from sklearn.metrics import log_loss
```

```python
df_train = pd.read_csv("/kaggle/input/tabular-playground-series-may-2021/train.csv").drop(columns='id')
df_test = pd.read_csv("/kaggle/input/tabular-playground-series-may-2021/test.csv").drop(columns='id')
submission = pd.read_csv('../input/tabular-playground-series-may-2021/sample_submission.csv')
```

```python
y = df_train.target

l, c = np.unique(y, return_counts=True)
c = c / c.sum()
y_pred = [c for _ in y]
log_loss(y_pred=y_pred, y_true=y, labels=l)
```

```python
y = df_train.target

l, c = np.unique(y, return_counts=True)
y_pred = [[0.25, 0.25, 0.25, 0.25] for _ in y]
log_loss(y_pred=y_pred, y_true=y, labels=l)
```
