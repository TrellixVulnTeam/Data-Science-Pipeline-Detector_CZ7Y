### To address imbalanced classification - over-sampling and under-sampling technique used.
### Smote - Synthetic Minority Oversampling Technique is a kind of approach, tried to explain this with the dataset of TBS May 2021. Hope this will be helpful for beginners to understand.

```python
from imblearn.combine import SMOTETomek
from collections import Counter

import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
dataset = pd.read_csv('/kaggle/input/tabular-playground-series-may-2021/train.csv')
```

```python
X = dataset.iloc[:, 1:51]
y = dataset.iloc[:, 51:]
```

### Below plot shows the imbalanced target variable 

```python
sns.countplot(x=y['target'], data=y)
```

```python
print(X.shape, y.shape)
print(Counter(y.target))
```

### Smote - Synthesizing elements for the minority class. It works by randomly picking a point from the minority class and computing K-Nearest Neighbors.
### Tomek - Pairs of very close instances, but of opposite classes. Removing the instances of majority class.
### Combining this over-sampling and under-sampling technique, we can balance the target class.
### Imblearn library provides a combination class SmoteTomek 

```python
smo_tek = SMOTETomek(random_state=0)
X_smotek, y_smotek = smo_tek.fit_resample(X, y)
```

### Under and Over sampling applied for the data using SmoteTomek
### Shape shows the increment of additional sampling data
### Counter shows the balanced target data

```python
print(X_smotek.shape, y_smotek.shape)
print(Counter(y_smotek.target))
```

```python
sns.countplot(x='target', data=y_smotek)
```
