## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
from xgboost import XGBClassifier

import warnings
warnings.filterwarnings('ignore')
```

## Importing the dataset

```python
dataset = pd.read_csv('/kaggle/input/tabular-playground-series-may-2021/train.csv')
test_data = pd.read_csv('/kaggle/input/tabular-playground-series-may-2021/test.csv')
submission_data = pd.read_csv('/kaggle/input/tabular-playground-series-may-2021/sample_submission.csv')
```

## Data Preprocessing

```python
#dataset.isna().sum()
#test_data.isna().sum()
```

```python
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values

test_data = test_data.iloc[:,1:].values
```

```python
le = LabelEncoder()
y = le.fit_transform(y)
```

## Splitting the Training dataset

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

## Train using XGBoost

```python
xgbc = XGBClassifier(random_state = 0, eval_metric = 'mlogloss', max_depth = 3)
xgbc.fit(X_train, y_train)
```

## Confusion Matrix

```python
y_pred = xgbc.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

## Prediction of Test Data

```python
pred_prob = xgbc.predict_proba(test_data)
print(np.round(pred_prob,2))
pred_prob = np.round(pred_prob,2)
```

## Submission

```python
submission_data.iloc[:,1:] = pred_prob
pd.DataFrame(submission_data).head(5)
```

```python
submission_data.to_csv('submission_xgbc.csv', index = False)
```
