```python
# Load the dataset in a Pandas dataframe.
train_df = pd.read_csv('../input/tabular-playground-series-may-2021/train.csv')
test_df = pd.read_csv('../input/tabular-playground-series-may-2021/test.csv')
```

```python
# Convert the dataset into a TensorFlow dataset.
train_ds = tfdf.keras.pd_dataframe_to_tf_dataset(train_df, label="target")
test_ds = tfdf.keras.pd_dataframe_to_tf_dataset(test_df)
```

```python
model_2 = tfdf.keras.GradientBoostedTreesModel(
    num_trees=1000,
    early_stopping_num_trees_look_ahead = 35,
    growing_strategy="BEST_FIRST_GLOBAL",
    max_depth=50,
    min_examples =10,
    split_axis="SPARSE_OBLIQUE",
    verbose = True,
    validation_ratio= 0.3,
    early_stopping = "LOSS_INCREASE"	
    )
```

```python
model_2.fit(x=train_ds)
```

```python
pred_2 = model_2.predict(test_ds)
```

```python
submission = pd.read_csv('../input/tabular-playground-series-may-2021/sample_submission.csv')
submission_df = pd.DataFrame(pred_2)
submission_df.columns = ['Class_1', 'Class_2', 'Class_3', 'Class_4']
submission_df['id'] = submission['id']
submission_df = submission_df[['id', 'Class_1', 'Class_2', 'Class_3', 'Class_4']]
submission_df.to_csv("submission_tfdf_V2.csv", index=False)
display(submission_df.head())
```
