```python
import numpy as np
import pandas as pd
import os

from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.layers import Flatten
from keras.callbacks import EarlyStopping
from keras.layers.convolutional import Conv1D
from keras.layers.convolutional import MaxPooling2D
from keras.layers.normalization import BatchNormalization
from keras.utils import np_utils

from sklearn.preprocessing import LabelBinarizer
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('../input/tabular-playground-series-may-2021/train.csv')
test = pd.read_csv('../input/tabular-playground-series-may-2021/test.csv')
```

```python
print(train.shape)
train.head()
```

```python
print(test.shape)
test.head()
```

```python
X = train.iloc[:,1:-1]
y0 = train.iloc[:,-1]
Test = test.iloc[:,1:]
```

```python
print(X.shape)
print(y0.shape)
print(Test.shape)
```

```python
X.head()
```

```python
y0
```

```python
binencoder = LabelBinarizer()
y = binencoder.fit_transform(y0)
y
```

```python
X_images = X.values.reshape(-1,5,10)
test_images = Test.values.reshape(-1,5,10)

print(X_images.shape)
print(test_images.shape)
```

```python
plt.imshow(X_images[5])
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_images, y, test_size = 0.2, random_state=90)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
X_train = X_train/255
X_test = X_test/255
X_train = X_train.reshape(-1,5,10).astype('float32')
X_test = X_test.reshape(-1,5,10).astype('float32')
```

```python
test_images2 = test_images/255
test = test_images2.reshape(-1,5,10).astype('float32')
```

```python
model = Sequential()

model.add(Conv1D(32,(2),input_shape=(5,10),activation='relu'))
model.add(BatchNormalization())
model.add(Conv1D(128,(2),activation='relu'))
model.add(Dropout(0.3))
model.add(Conv1D(512,(2),activation='relu'))
model.add(Dropout(0.3))
model.add(Flatten())
model.add(Dense(128,activation='relu'))
model.add(Dense(64,activation='relu'))
model.add(Dense(4,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
```

```python
result = model.fit(X_train,y_train,validation_split=0.2,epochs=300,batch_size=64,verbose=2)
```

```python
history_df = pd.DataFrame(result.history)
history_df.loc[:,['loss','val_loss','accuracy','val_accuracy']].plot()
```

```python
y_pred = model.predict(test)
print(y_pred[0])
```

```python
y_pred_df=pd.DataFrame(y_pred)
y_pred_df
```
