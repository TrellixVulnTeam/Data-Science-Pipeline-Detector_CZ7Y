```python
df_main.set_index("id",inplace=True)
```

```python
col_list=df_main.columns.tolist()
sparasity_dict={}
for col in col_list:
    sparasity_dict[col]=df_main[(df_main[col]==0)].shape[0]/df_main.shape[0]
```

```python
sparasity_dict
```

```python
pipe = Pipeline([('pca', PCA(n_components=4)), ('GBR', GradientBoostingClassifier())])
X,y=df_main.drop(["target"],1),df_main["target"]
pipe.fit(X,y)
```

```python
df_test=pd.read_csv("../input/tabular-playground-series-may-2021/test.csv")
predictions=pipe.predict(df_test.set_index("id"))
```

```python
df_sub=pd.DataFrame()
df_sub["id"]=df_test.index
df_sub["Predictions"]=predictions
df_sub.to_csv("submission.csv")
```
