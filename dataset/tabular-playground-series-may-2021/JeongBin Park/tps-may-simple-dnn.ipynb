## Data
* train.csv - the training data, one product (id) per row, with the associated features (feature_*) and class label (target)
* test.csv - the test data; you must predict the probability the id belongs to each class
* sample_submission.csv - a sample submission file in the correct format

## Import library and Load Dataset

```python
import numpy as np
import pandas as pd
from keras import layers, models
import matplotlib.pyplot as plt

train = pd.read_csv("../input/tabular-playground-series-may-2021/train.csv")
test = pd.read_csv("../input/tabular-playground-series-may-2021/test.csv")
```

```python
train.head()
```

```python
test.head()
```

#### The data have 52columns (include target) and 100000rows
#### I don't know about features (columns featrue_0 ~ feature_49). So I just assume that all features are meaningful data for predict target

```python
train = train.drop('id',axis=1)
test = test.drop('id',axis=1)

X = train.drop('target',axis=1)
y = train['target']
y = pd.get_dummies(y)
```

```python
result = pd.read_csv("../input/tabular-playground-series-may-2021/sample_submission.csv")
result.head()
```

```python
y.head()
```

#### I just drop 'id' column and Change the target's shape like sample_sumbission.csv's shape

## Modeling  - Deep Learning (DNN by keras)

```python
model = models.Sequential()
model.add(layers.Dense(50, input_shape=X.shape, activation='relu', name="Hidden-1"))
model.add(layers.Dropout(0.25))
model.add(layers.Dense(25, activation='relu'))
model.add(layers.Dropout(0.25))
model.add(layers.Dense(10, activation='softmax'))
model.add(layers.Dropout(0.1))
model.add(layers.Dense(4, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
history = model.fit(X, y, epochs=30, batch_size=256, validation_split=0.3)
```

```python
plt.figure(figsize=(15,8))
plt.plot(history.history['loss'],lw=4)
plt.plot(history.history['val_loss'],lw=4)
plt.title('Model Loss',fontsize=20)
plt.ylabel('Loss',fontsize=15)
plt.xlabel('Epoch',fontsize=15)
plt.legend(['Train','Test'])
plt.show()
```

## Predict ! 

```python
pred = model.predict(test)
result[['Class_1','Class_2','Class_3','Class_4']] = pred
result.head(10)
```

```python
result.to_csv("result.csv",index=None)
```
