## Collection of various Feature Reduction Methods 

```python
from sklearn.decomposition import PCA, SparsePCA
```

```python
train_org = pd.read_csv('../input/tabular-playground-series-may-2021/train.csv')
print(train_org.shape)

test_org = pd.read_csv('../input/tabular-playground-series-may-2021/test.csv')
print(test_org.shape)

df_dae = pd.read_csv('../input/dae-te-le-tps0521/df_dae.csv')
print(df_dae.shape)

df_le = pd.read_csv("../input/dae-te-le-tps0521/df_le.csv")
print(df_le.shape)

df_te = pd.read_csv("../input/dae-te-le-tps0521/df_te.csv")
print(df_te.shape)

y = pd.read_csv("../input/dae-te-le-tps0521/y.csv")
print(y.shape)
```

```python
X = pd.concat([df_dae, df_le, df_te], axis = 1)[:100000]
print(X.shape)

X_tst = pd.concat([df_dae, df_le, df_te], axis = 1)[100000:]
print(X_tst.shape)
```

PCA Standard

```python
pca = PCA(n_components = 40)

X_pca = pca.fit_transform(X)

X_tst_pca = pca.transform(X_tst)

print(sum(pca.explained_variance_ratio_))
```

```python
print(X_pca.shape)

print(X_tst_pca.shape)


np.save('X_pca.npy', X_pca)
np.save('X_tst_pca.npy', X_tst_pca)

#X_pca.to_csv("X_pca.csv", index = False)
#X_tst_pca.to_csv("X_tst_pca.csv", index = False)
```

Random Forest Feature Selection

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(random_state=1, max_depth=10)
#df=pd.get_dummies(df)
model.fit(X,y)


features = X.columns
importances = model.feature_importances_
indices = np.argsort(importances)[-19:]  # top 10 features
plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()

from sklearn.feature_selection import SelectFromModel
feature = SelectFromModel(model)
Fit = feature.fit_transform(X,y)
```

```python
from collections import Counter 
from imblearn.over_sampling import SMOTE

counter = Counter(np.array(y['target']))
print(counter)
# transform the dataset
oversample = SMOTE()
X_smote, y_smote = oversample.fit_resample(X, y)
# summarize the new class distribution
```
