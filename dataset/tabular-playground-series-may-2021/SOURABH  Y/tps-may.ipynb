```python
import numpy as np 
import pandas as pd 
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
import imblearn
from imblearn.over_sampling import SMOTE
```

```python
data=pd.read_csv('../input/tabular-playground-series-may-2021/train.csv')
```

```python
X = pd.DataFrame(data.drop(["target","id"], axis = 1))
lencoder = LabelEncoder()
y = pd.DataFrame(lencoder.fit_transform(data['target']), columns=['target'])
```

```python
y
```

```python
len(y[y['target']==3])
```

```python
oversample = SMOTE()
X, y = oversample.fit_resample(X, y)
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train
```

```python
y_train
```

```python
test=pd.read_csv('../input/tabular-playground-series-may-2021/test.csv')
```

```python
test = pd.DataFrame(test.drop("id", axis = 1))
```

```python
test.shape
```

```python
import xgboost as xgb
```

```python
model = xgb.XGBClassifier(objective='multi:softprob',use_label_encoder=False)
model.fit(X_train, y_train)
```

```python
y_pred=model.predict_proba(X_test)
```

```python
y_pred
```

```python
pred2 = model.predict_proba(test)
pred2
```

```python
sub = pd.read_csv("/kaggle/input/tabular-playground-series-may-2021/sample_submission.csv")

predictions = pd.DataFrame(pred2, columns = ["Class_1", "Class_2", "Class_3", "Class_4"])
predictions['id'] = sub['id']
predictions.to_csv("subm5.csv", index = False)
```
