This is a synthetic dataset that is based on a real dataset. These datasets are based on Kaggle's May 2021 tabular competition.

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

Read the files

```python
#Reading train file:
train = pd.read_csv('/kaggle/input/tabular-playground-series-may-2021/train.csv')
#Reading test file:
test = pd.read_csv('/kaggle/input/tabular-playground-series-may-2021/test.csv')
#reading sample submission
submission = pd.read_csv('/kaggle/input/tabular-playground-series-may-2021/sample_submission.csv')
```

Analyse target

```python
import matplotlib.pyplot as plt

train.groupby('target').target.count().plot.bar(ylim=0)
plt.show()
```

One hot encode target

```python
target = pd.get_dummies(train.target)
target.columns=['Class_1', 'Class_2', 'Class_3', 'Class_4']
target
```

Drop target from train

```python
train.drop('target', axis = 1, inplace = True)
train
```

Combine train and test

```python
combi = train.append(test)
combi
```

Drop ID from combi

```python
combi.drop('id', axis = 1, inplace = True)
combi
```

Normalise combi

```python
combi = (combi.max() - combi) / (combi.max() - combi.min())
combi
```

Define X and y

```python
y = target
X = combi[: len(train)]
X_test = combi[len(train) :]
```

```python
y
```

```python
X
```

```python
X_test
```

Split into training and validation

```python
#split train set for testing
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.10, random_state=1)
X_train.shape, y_train.shape, X_val.shape, y_train.shape, X_test.shape
```

Select model

```python
from sklearn.multioutput import MultiOutputRegressor
from sklearn import linear_model

model = MultiOutputRegressor(linear_model.BayesianRidge(n_iter=10000)).fit(X_train, y_train)
model.score(X_train, y_train)
```

Predict on validation set

```python
y_pred = model.predict(X_val)
model.score(X_val, y_val)
```

```python
y_pred
```

Predict on test set

```python
predictions = model.predict(X_test)
predictions.shape
```

```python
predictions[predictions < 0] = 0
```

Prepare submission

```python
submission.Class_1 = predictions[:, 0]
submission.Class_2 = predictions[:, 1]
submission.Class_3 = predictions[:, 2]
submission.Class_4 = predictions[:, 3]
```

```python
submission.to_csv('submission.csv', index=False)
submission = pd.read_csv("submission.csv")
submission
```
