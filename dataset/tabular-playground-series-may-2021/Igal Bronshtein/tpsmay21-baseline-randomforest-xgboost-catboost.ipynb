## Baseline Random Forest XGBoost CatBoost

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
import catboost
from catboost import CatBoostClassifier
from sklearn.metrics import log_loss
import warnings
warnings.filterwarnings(action='ignore', category=UserWarning)

df = pd.read_csv('../input/tabular-playground-series-may-2021/train.csv')
```

```python
df.head(1)
```

### Random Forest

```python
clf = RandomForestClassifier(random_state = 0)

X = df.drop("target", axis=1)
y = df["target"].astype('category').cat.codes
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 0)
```

```python
clf.fit(X_train, y_train);
```

```python
predictions = clf.predict_proba(X_test)
predictions
```

```python
clf.score(X_test, y_test) 
```

```python
log_loss_result = log_loss(y_test, predictions)
round(log_loss_result, 5)
```

```python
test_df = pd.read_csv('../input/tabular-playground-series-may-2021/test.csv')
submission_df = pd.read_csv('../input/tabular-playground-series-may-2021/sample_submission.csv')
submission_df.head(2)
```

## XGBoost

```python
clf = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state = 0)
clf.fit(X_train, y_train);
```

```python
predictions = clf.predict_proba(X_test).copy()
predictions
```

```python
clf.score(X_test, y_test) 
```

```python
log_loss_result = log_loss(y_test, predictions)
round(log_loss_result, 5)
```

## CatBoost

```python
clf = CatBoostClassifier(random_state = 0)
clf.fit(X_train, y_train, logging_level = 'Silent');
```

```python
predictions = clf.predict_proba(X_test)
print(predictions)
```

```python
clf.score(X_test, y_test) 
```

```python
log_loss_result = log_loss(y_test, predictions)
round(log_loss_result, 5)
```

#### CatBoost gives the best score.
