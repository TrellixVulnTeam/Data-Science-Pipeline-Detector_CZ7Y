```python
import pandas as pd

df = pd.read_csv('../input/tabular-playground-series-may-2021/train.csv')
emojis_df = df.copy()
emojis_df.head(2)
```

By changing the feature names to emojis it would be possible to see more columns without scrolling to the right. I choose flags as their names are in ABC order. So would be more easy for me, remember the order of features.

```python
emojis_df.columns = ["🆔", "🇦🇫", "🇦🇬", "🇦🇱", "🇦🇲", "🇦🇶", "🇦🇷", "🇦🇸", "🇦🇹", "🇦🇼", "🇦🇽", "🇦🇿", "🇧🇦", "🇧🇧",
              "🇧🇩", "🇧🇪", "🇧🇫", "🇧🇬", "🇧🇭", "🇧🇮", "🇧🇯", "🇧🇱", "🇧🇳", "🇧🇴", "🇧🇶", "🇧🇷", "🇧🇸", "🇧🇹",
              "🇧🇻", "🇧🇼", "🇧🇾", "🇧🇿", "🇨🇦", "🇨🇨", "🇨🇩", "🇨🇫", "🇨🇬", "🇨🇭", "🇨🇮", "🇨🇱", "🇨🇲", "🇨🇳",
              "🇨🇴", "🇨🇵", "🇨🇷", "🇨🇺", "🇨🇻", "🇨🇼", "🇨🇽", "🇨🇾", "🇨🇿", "🎯"]

pd.set_option('display.max_columns', None)
emojis_df.head(2)
```

Converting "target" to emojis 

```python
def convert_target_to_emojis(x):   
    return { 'Class_1' : "👛", 'Class_2': "👜", 'Class_3': "👝", "Class_4": "🎒"}[x]

for i in emojis_df.index:
    current_item = emojis_df.at[i, "🎯"]
    emojis_df.at[i, "🎯"] = convert_target_to_emojis(current_item)

pd.set_option('display.max_columns', 20)    
emojis_df.head(14)
```

There a lot of zeros, will replace them with emojis.

```python
emojis_f_names = ["🇦🇫", "🇦🇬", "🇦🇱", "🇦🇲", "🇦🇶", "🇦🇷", "🇦🇸", "🇦🇹", "🇦🇼", "🇦🇽", "🇦🇿", "🇧🇦", "🇧🇧",
              "🇧🇩", "🇧🇪", "🇧🇫", "🇧🇬", "🇧🇭", "🇧🇮", "🇧🇯", "🇧🇱", "🇧🇳", "🇧🇴", "🇧🇶", "🇧🇷", "🇧🇸", "🇧🇹",
              "🇧🇻", "🇧🇼", "🇧🇾", "🇧🇿", "🇨🇦", "🇨🇨", "🇨🇩", "🇨🇫", "🇨🇬", "🇨🇭", "🇨🇮", "🇨🇱", "🇨🇲", "🇨🇳",
              "🇨🇴", "🇨🇵", "🇨🇷", "🇨🇺", "🇨🇻", "🇨🇼", "🇨🇽", "🇨🇾", "🇨🇿"]

for i in emojis_f_names:
    emojis_df[i] = emojis_df[i].apply(lambda x: "🔵" if x == 0 else "🟣")
```
