```python
import sklearn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sample_submission = pd.read_csv('../input/tabular-playground-series-may-2021/sample_submission.csv')
test = pd.read_csv("../input/tabular-playground-series-may-2021/test.csv")
train = pd.read_csv("../input/tabular-playground-series-may-2021/train.csv")
```

```python
from pycaret.classification import *

grid=setup(data=train, target='target', verbose=True, fix_imbalance=True,
           experiment_name='adv1', fold=5)
```

```python
topmodels = compare_models(n_select = len(models()))
```

```python
a = topmodels[0]
b = topmodels[1]
c = topmodels[2]
d = topmodels[3]
e = topmodels[4]
```

```python
blender = blend_models(estimator_list=[a, b, c, d, e], method='auto', fold=5)
```

```python
subm = predict_model(blender,data = test, raw_score = True)
```

```python
subm
```

```python
final_sub = subm[['id','Score_Class_1','Score_Class_2','Score_Class_3','Score_Class_4']]
```

```python
final_sub.rename(columns={'Score_Class_1':'Class_1',
              'Score_Class_2':'Class_2',
              'Score_Class_3':'Class_3','Score_Class_4':'Class_4'}, inplace=True)
final_sub.head()
```
