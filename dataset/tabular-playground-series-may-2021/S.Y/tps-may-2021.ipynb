```python
def code_transform(x):
    if x == 'Class_1':
        y = 0
    elif x == 'Class_2':
        y = 1
    elif x == 'Class_3':
        y = 2
    elif x == 'Class_4':
        y = 3
    return y

train_drop_id['target'] = train_drop_id['target'].apply(lambda x: code_transform(x))
```

```python
from sklearn.model_selection import train_test_split

x_train_b, x_test_b = train_test_split(train_drop_id, test_size=0.2, random_state=1)

y_train = x_train_b['target'].ravel()
x_train = x_train_b.copy().drop(['target'], axis=1)

y_test = x_test_b['target'].ravel()
x_test = x_test_b.copy().drop(['target'], axis=1)


# for cpl in x_train.columns:
#     l=[]
#     for i in x_train[str(col)]:
#         l.append(np.log(1+i))
            
#     x_train[str(col)] = l
    
#     l=[]
#     for i in x_test[str(col)]:
#         l.append(np.log(1+i))
            
#     x_test[str(col)] = l
```

```python
import lightgbm as lgb
lgb.LGBMClassifier()
```

```python
train_data = lgb.Dataset(x_train, label=y_train)
eval_data = lgb.Dataset(x_test, label=y_test, reference= train_data)
```

```python
params = {
         'task': 'train',
         'boosting_type': 'gbdt',
         'objective': 'multiclass',
         'num_class': 4,
         'verbose': 2
         }


gbm = lgb.train(params,
                train_data,
                valid_sets=eval_data,
                num_boost_round=100,
                verbose_eval=5
                )
```

```python
preds = gbm.predict(x_test)
preds
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

y_pred = []

for x in preds:
    y_pred.append(np.argmax(x))
accuracy_score(y_test, y_pred)
```

```python
gbm_predict = gbm.predict(target)
gbm_predict
```
