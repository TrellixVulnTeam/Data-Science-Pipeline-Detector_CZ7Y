```python
import pandas as pd
from xgboost import XGBClassifier
```

```python
train_df = pd.read_csv('../input/tabular-playground-series-may-2021/train.csv')
train_df
```

```python
X = train_df.drop(['id', 'target'], axis=1)
X
```

```python
y = train_df['target']
y
```

```python
xg_model = XGBClassifier(random_state=42, tree_method='gpu_hist', use_label_encoder=True)
xg_model = xg_model.fit(X, y)
xg_model
```

```python
test_df = pd.read_csv('../input/tabular-playground-series-may-2021/test.csv')
test_df
```

```python
X_test = test_df.drop(['id'], axis=1)
X_test
```

```python
xg_model.predict_proba(X_test)
```

```python
submission_df = pd.DataFrame(xg_model.predict_proba(X_test), columns=xg_model.classes_)
submission_df
```

```python
submission_df.insert(0, 'id', test_df['id'])
```

```python
submission_df
```

```python
submission_df.to_csv('submission.csv', index=0)
```
