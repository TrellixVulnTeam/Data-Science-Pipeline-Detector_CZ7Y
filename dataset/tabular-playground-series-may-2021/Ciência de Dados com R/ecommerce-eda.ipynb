```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

### Carregar a Base de Treino

```python
train = pd.read_csv('/kaggle/input/tabular-playground-series-may-2021/train.csv')

X_train = train.drop(columns=['target'])
```

```python
train.head()
```

```python
cols = train.columns[1:-1]
cols
```

```python
def kdechart(col):
    plt.figure(figsize=(5,5))
    sns.histplot(x=train[col], hue=train['target'])
    plt.title("Distribution of "+ col,fontsize=15)
    #plt.legend(['Class_1','Class_2','Class_3','Class_4'])
    plt.legend(train['target'])
    plt.show();

for col in cols:
    kdechart(col)
```
