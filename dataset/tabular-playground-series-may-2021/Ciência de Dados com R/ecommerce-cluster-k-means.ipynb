```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.cluster import KMeans, AgglomerativeClustering
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt
import seaborn as sns

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

# K-MEANS

```python
X = pd.read_csv('/kaggle/input/tabular-playground-series-may-2021/train.csv')

# YOUR CODE HERE: Define a list of the features to be used for the clustering
features = X.columns[1:-1]

X_features = X.loc[:, features]

kmeans = KMeans(n_clusters = 4,n_init=10)

X['Cluster'] = kmeans.fit_predict(X_features)

X.head()
X_features.head()
```

```python
data = X.loc[:,['target','Cluster']]
plt.figure(figsize=(20,8))
sns.histplot(data = X[X['target']=='Class_1']['Cluster'], kde= False, color = 'Red')
sns.histplot(data = X[X['target']=='Class_2']['Cluster'], kde= False)
sns.histplot(data = X[X['target']=='Class_3']['Cluster'], kde= False, color = 'Yellow')
sns.histplot(data = X[X['target']=='Class_4']['Cluster'], kde= False)
```

```python
plt.figure(figsize=(20,8))
sns.scatterplot('feature_0','feature_1',data=X,hue='Cluster')
```
