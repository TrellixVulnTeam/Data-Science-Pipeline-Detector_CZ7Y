```python
import pandas as pd
```

```python
train_X = pd.read_csv('../input/tabular-playground-series-may-2021/train.csv')
test_X = pd.read_csv('../input/tabular-playground-series-may-2021/test.csv')
```

```python
# train_X.head()
print(set(train_X['target']))
```

```python
test_X.head()
```

```python
print(train_X.shape,test_X.shape)
```

```python
target = train_X['target']
train_X.drop(['target'],axis='columns',inplace=True)
train_X.drop(['id'],axis='columns',inplace=True)
```

```python
test_id = test_X['id']
test_X.drop(['id'],axis='columns',inplace=True)
```

```python
print(train_X.shape,test_X.shape,target.shape)
```

```python
# check for NAN
train_X.isnull().sum()
```

```python
# check for total sum == 0
train_X.sum() == 0
```

```python
from sklearn.model_selection import train_test_split

x_train,y_train,x_test,y_test = train_test_split(train_X,target,test_size=0.5,random_state=42)
```

```python
print(x_train.shape,y_train.shape,x_test.shape,y_test.shape)
```

```python
from sklearn import svm

svm_clf = svm.SVC(kernel='rbf',C=1,gamma=1.0,probability=True).fit(x_train,x_test)
```

```python
y_pred = svm_clf.predict(y_train)
```

```python
from sklearn import metrics

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
final = pd.DataFrame(svm_clf.predict_proba(test_X),columns=svm_clf.classes_)
```
