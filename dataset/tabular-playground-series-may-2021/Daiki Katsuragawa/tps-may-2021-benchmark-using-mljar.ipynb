```python
import pandas as pd
from supervised.automl import AutoML
from supervised.preprocessing.eda import EDA

train = pd.read_csv('../input/tabular-playground-series-may-2021/train.csv')
test = pd.read_csv('../input/tabular-playground-series-may-2021/test.csv')
label = 'target'

X_train,y_train = train.drop([label],axis=1),train[label]
EDA.extensive_eda(X_train,y_train,save_path="/")
```

```python
eval_metric = 'logloss'
total_time_limit = 600
automl = AutoML(
    mode='Compete',
    eval_metric=eval_metric,
    total_time_limit=total_time_limit
)
automl.fit(X_train,y_train)
automl.report()
```

```python
submission = pd.read_csv('../input/tabular-playground-series-may-2021/sample_submission.csv')
columns = ['Class_1', 'Class_2', 'Class_3', 'Class_4']
predictions = pd.DataFrame(automl.predict_proba(test), columns=columns)
for column in columns:
    submission[column] = predictions[column]
submission.to_csv('submission.csv', index=False)
submission.head()
```
