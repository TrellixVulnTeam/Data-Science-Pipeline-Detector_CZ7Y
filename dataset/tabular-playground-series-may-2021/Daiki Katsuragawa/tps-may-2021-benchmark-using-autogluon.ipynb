```python
import pandas as pd
from autogluon.tabular import TabularDataset, TabularPredictor

train = TabularDataset('../input/tabular-playground-series-may-2021/train.csv')
test = TabularDataset('../input/tabular-playground-series-may-2021/test.csv')

label = 'target'
time_limit = 3600 * 2
predictor = TabularPredictor(label=label,learner_kwargs={'ignored_columns': ['id']}).fit(train_data=train,time_limit=time_limit)
```

```python
submission = pd.read_csv('../input/tabular-playground-series-may-2021/sample_submission.csv')
predictions = predictor.predict_proba(test)
columns = ['Class_1', 'Class_2', 'Class_3', 'Class_4']
for column in columns:
    submission[column] = predictions[column]
submission.to_csv('submission.csv', index=False)
submission.head()
```
