```python
import pandas as pd
#CSV = "../input/tabular-playground-series-may-2021/train.csv"
CSV = "../input/titanic/train.csv"
train = pd.read_csv(CSV)
print(train.shape)
display(train.head(3))
```

# pandas_profiling

```python
%%time
# csvのサマリーhtml出力
import pandas_profiling
profile = train.profile_report()
profile.to_file("profile.html")
profile
```

# sweetviz

```python
%%time
# csvのサマリーhtml出力。pandas_profilingより処理速度早い
import sweetviz as sv
sweet_report = sv.analyze(train)
sweet_report.show_html('sweet_report.html')
sweet_report.show_notebook()
```

```python
# sweetvizは2つのデータセット（train／test）比較できる
if "titanic" in CSV :
    test = pd.read_csv("../input/titanic/test.csv")
    compare_report = sv.compare([train, "train"], [test, "test"], "Survived")  # 最後の引数には目的変数を指定
    compare_report.show_html('compare_report.html')
    compare_report.show_notebook()
```

# autoviz

```python
%%time
# csvをいろんなグラフで可視化
from autoviz.AutoViz_Class import AutoViz_Class
autoviz = AutoViz_Class().AutoViz(CSV)
```

# dtale

```python
# dtaleはテーブルをブラウザ上でいじくるライブラリ。ピボットテーブル作ったりできる
# kaggleでは動かず
import dtale
dtale.show(train)
```
