```python
import numpy as np
import pandas as pd
import os
import folium
from folium import plugins
```

```python
names=[]
paths=[]
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        if filename[-4:]=='.csv':
            names+=[filename[0:-4]]
            paths+=[os.path.join(dirname, filename)]
```

```python
print(names)
```

```python
for i in range(len(paths)):
    names[i]=pd.read_csv(paths[i])
    print(str(i)+'. '+paths[i].split('/')[-1][0:-4])
    display(names[i])
    print()
```

```python
common_names=names[3]['common_name'].unique().tolist()
print(len(common_names))
```

```python
print('All Data')
data=names[3][['latitude','longitude']].reset_index(drop=True)
eq_map = folium.Map(location=[20,0],tiles='Stamen Terrain',zoom_start=1,min_zoom=2.0)
eq_map.add_child(plugins.HeatMap(data))
eq_map
```

```python
i=10
item=common_names[i]
print(item)
data=names[3][names[3]['common_name']==item][['latitude','longitude']].reset_index(drop=True)
eq_map = folium.Map(location=[20,0],tiles='Stamen Terrain',zoom_start=1,min_zoom=2.0)
eq_map.add_child(plugins.HeatMap(data))
eq_map
```
