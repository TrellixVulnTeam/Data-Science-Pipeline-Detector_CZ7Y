```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
import json
import librosa
```

Completely random for now.

```python
with open('/kaggle/input/birdclef-2022/scored_birds.json') as bird_file:
    birds = json.load(bird_file)
```

```python
test_directory = '/kaggle/input/birdclef-2022/test_soundscapes/'
```

```python
test_files = os.listdir(test_directory)
```

```python
predictions = []
for test_file_name in test_files:
    ts, rate = librosa.load(test_directory + test_file_name)
    for segment in range(0, len(ts), rate*5):
        for bird in birds:
            row_id = test_file_name[:-4] + '_' + bird + '_' + str(int((segment + (rate * 5)) / (rate)))
            target = bool(np.random.randint(2))
            predictions.append([row_id, target])
```

```python
submission = pd.DataFrame(predictions, columns=['row_id', 'target'])
```

```python
submission.to_csv('submission.csv', index=False)
```
