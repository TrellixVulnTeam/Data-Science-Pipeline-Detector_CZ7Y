```python
from google.cloud import storage
client = storage.Client(project="birdclef-2022")

from pathlib import Path
for blob in client.list_blobs("birdclef-2022", prefix="processed/classify-nn/0.13.0-202205222347"):
    name = Path(blob.name).relative_to("processed")
    name.parent.mkdir(parents=True, exist_ok=True)
    blob.download_to_filename(name)
    print(f"downloaded {name}")
```
