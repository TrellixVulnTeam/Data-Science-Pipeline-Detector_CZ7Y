# Convert Audio to Frequency

```python
import soundfile as sf
freq,rate=sf.read('../input/birdclef-2022/train_audio/afrsil1/XC125458.ogg')
import plotly.express as px
import numpy as np
px.line(x=np.array(list(range(len(freq)))),y=freq)
```

# Remove the noise from the sound

```python
!pip install noisereduce
```

```python
import noisereduce as nr
reduced_noise=nr.reduce_noise(y=freq,sr=rate)
```

```python
px.line(x=np.array(list(range(len(reduced_noise)))),y=reduced_noise)
```

```python
import wavio
wavio.write("myfile.wav",reduced_noise,rate,sampwidth=2)
```

# Convert to spectogram using tfio [chanage params here to reduce memory usage]

```python
!pip install tensorflow-io
```

```python
import tensorflow_io as tfio
import matplotlib.pyplot as plt
# Convert to spectrogram
spectrogram = tfio.audio.spectrogram(
    reduced_noise, nfft=3600, window=256, stride=256)
import tensorflow as tf
plt.figure()
plt.imshow(tf.math.log(spectrogram).numpy())
```

# Finally a function to convert an ogg file to spectogram

```python
def read(pth,reduce_noise):
    freq,rate=sf.read(pth)
    if reduce_noise:
        freq=nr.reduce_noise(y=freq,sr=rate)
    # Convert to spectrogram
    spectrogram = tfio.audio.spectrogram(
    reduced_noise, nfft=3600, window=256, stride=256)
    return tf.math.log(spectrogram).numpy()
plt.imshow(read('../input/birdclef-2022/train_audio/afrsil1/XC175522.ogg',True))
```
