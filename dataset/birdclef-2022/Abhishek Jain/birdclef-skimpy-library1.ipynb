```python
import os
import pandas as pd
import numpy as np
from pathlib import Path
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.patches as mpl_patches
import matplotlib.pylab as pylab
params = {'legend.fontsize': 'x-large',
          'figure.figsize': (15, 8),
         'axes.labelsize': 'x-large',
         'axes.titlesize':'x-large',
         'xtick.labelsize':'x-large',
         'ytick.labelsize':'x-large',
         'text.color' : "green",
          
         }
pylab.rcParams.update(params)

import torch
import torchaudio
from torch.utils.data.dataset import Dataset
from torch.utils.data.dataloader import DataLoader
from torchvision.models import resnet34
import torch
import torch.nn as nn
import torch.optim as optim
import librosa
from tqdm.notebook import tqdm

from colorama import Fore, Back, Style
y_ = Fore.YELLOW
r_ = Fore.RED
g_ = Fore.GREEN
b_ = Fore.BLUE
m_ = Fore.MAGENTA
c_ = Fore.CYAN
sr_ = Style.RESET_ALL

import warnings
warnings.filterwarnings('ignore')
plt.style.use('fivethirtyeight')
import IPython.display as ipd
from IPython.display import display, HTML
import plotly.express as px
print('Execution success')
```

```python
train = pd.read_csv("../input/birdclef-2022/train_metadata.csv")
test = pd.read_csv("../input/birdclef-2022/test.csv")
sub = pd.read_csv("../input/birdclef-2022/sample_submission.csv")
taxonomy = pd.read_csv( '../input/birdclef-2022/eBird_Taxonomy_v2021.csv')

print("{0}Number of rows in train data: {1}{2}\n{0}Number of columns in train data: {1}{3}"
      .format(y_,r_,train.shape[0],train.shape[1]))
print("{0}Number of rows in test data: {1}{2}\n{0}Number of columns in train data: {1}{3}"
      .format(m_,r_,test.shape[0],test.shape[1]))
```

```python
print(f"{m_}number of unique primary label: {r_}{train.primary_label.nunique()}")
print(f"{c_}number of unique scientific name: {r_}{train.scientific_name.nunique()}")
```

```python
g = sns.countplot(data=taxonomy, x = "CATEGORY")
plt.title("Record count per Category")
for p in g.patches:
    g.annotate(format(p.get_height(), '.0f'),
               (p.get_x() + p.get_width() / 2., p.get_height()),
               ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')
plt.show()
```

```python
rec_dict = {}
labels_list = train["primary_label"].unique().tolist()[:10]
for i,spid in enumerate(labels_list):
    rec_id = train.loc[np.where(train["primary_label"] == spid)]["primary_label"].tolist()[0]
    audio_file = str(list(Path(os.path.join('../input/birdclef-2022/train_audio', labels_list[0]) ).glob('*.ogg'))[0])

#     print(f"{spid} : {rec_id}")
    rec_dict[spid] = rec_id
    fig, ax = plt.subplots()
    rec_path = audio_file #Path(train_path) / f"{rec_id}.flac"
    wf, sr = torchaudio.load(rec_path)
    labels = []
    handles = [mpl_patches.Rectangle((0, 0), 1, 1, fc="white", ec="white", lw=0, alpha=0)] * 4 
    labels.append(f"label: {spid}")
    labels.append(f"min of waveform: {np.around(wf.min(), 5)}")
    labels.append(f"max of waveform: {np.around(wf.max(), 5)}")
    labels.append(f"mean of waveform: {np.around(wf.mean(), 5)}")
    ax.legend(handles, labels, loc='best', fontsize='large', 
          fancybox=True, framealpha=0.7, 
          handlelength=0, handletextpad=0)
    
    plt.plot(wf.t().numpy())
    plt.title("Species wave form plot for 10 Category")
    plt.show()
```

```python
rec_dict = {}
display(HTML(f'<span style="color:green"> <h3>Sound of each Category </h3></span>'))

def get_species_audio():
    """
    get audio image for each of the species
    """
    for i,spid in enumerate(labels_list):
        rec_id = train.loc[np.where(train["primary_label"] == spid)]["primary_label"].tolist()[0]
    #     print(f"{spid} : {rec_id}")
        rec_dict[spid] = rec_id
        audio_file = str(list(Path(os.path.join('../input/birdclef-2022/train_audio', labels_list[0]) ).glob('*.ogg'))[0])
        wf, sr = torchaudio.load(audio_file)
        display(HTML(f'<span style="color:green"> Category id: {spid} </span>'))
        ipd.display(ipd.Audio(data=wf, rate=sr))
get_species_audio()
```

```python
fig = px.scatter_geo(
    train,
    lat="latitude",
    lon="longitude",
    color="common_name",
    width=1_000,
    height=500,
    title="BirdCLEF 2022 Training Data Location Plot",
)
fig.show()
```

```python
rec_dict = {}
display(HTML(f'<span style="color:green"> <h3>spectrogram of each species </h3></span>'))
def get_species_spectrogram():
    """
    get audio image for each of the species
    """
    for i,spid in enumerate(labels_list):
        rec_id = train.loc[np.where(train["primary_label"] == spid)]["primary_label"].tolist()[0]
        rec_dict[spid] = rec_id
        audio_file = str(list(Path(os.path.join('../input/birdclef-2022/train_audio', labels_list[0]) ).glob('*.ogg'))[0])
        wf, sr = torchaudio.load(audio_file)
        specgram = torchaudio.transforms.Spectrogram()(wf)

        fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(20, 4))
        ax.tick_params(axis='x', labelsize=10)
        ax.tick_params(axis='y', labelsize=10)

        cax = ax.matshow(
            specgram.log2()[0,:,:].numpy(),
            interpolation="nearest",
            aspect="auto",
            cmap=plt.cm.afmhot,
            origin="lower",
        )
        fig.colorbar(cax)

        plt.title(f"Spectrogram for Category id: {spid}")
        plt.show()
get_species_spectrogram()
```
