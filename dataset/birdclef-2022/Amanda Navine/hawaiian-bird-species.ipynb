## Waterfowl
There were once many species of ducks, geese, and goose-like birds inhabiting the Hawaiian Islands, many of them large and flightless. Because the archipelago lacked any endemic mammalian species, these waterfowl did not have any natural land predators until humans colonized the islands. Polynesians introduced mammalian predators such as pigs and dogs, and exploited these species for food, as well as for bones and feathers for making tools. This, followed by the arrival of Europeans and introduction of morepredators such as rats and cats, quickly drove most of the large and flightless species to extinction. Waterfowl are important to the islands’ wetland ecosystems, serving as seed and nutrient dispersers.

### Nēnē (*Branta sandvicensis*)
The state bird of Hawai'i, the Nēnē (also known as the Hawaiian Goose), is the smallest and most powerfully flighted of the geese that evolved in the islands, and also the only species living today. In contrast to many other waterfowl species, Nēnē are highly terrestrial and sedentary. They are herbivorous and important seed dispersers for grasses, herbs, and shrubs. Nēnē were recently down-listed from endangered to threatened under the Endangered Species Act, and are listed as near threatened by the IUCN. Today, they have an estimated population of over 3,000 individuals found on the islands of Hawai'i, Maui Nui, and Kaua'i, and their numbers are steadily increasing. However, despite the success conservation managers have had in boosting Nēnē numbers, they still face many threats to their survival such as habitat loss, introduced predators, habitat degradation by invasive species, and collisions with vehicles and wind turbines. Nēnē vocalizations are similar to those of their closest living relatives, Canada Geese, but softer, consisting of either low moaning “nay-nay” calls or shrill trumpeted calls.

## Hawaiian honeycreepers
Hawaiian honeycreepers are a group of small, colorful forest birds that exemplify adaptive radiation, having evolved into over 55 documented species from a single finch species that colonized the Hawaiian archipelago roughly seven million years ago. Honeycreeper species have diverse foraging strategies, from highly specialized snail eaters, to generalists who eat a large variety of food items. However, many honeycreepers are nectivorous, and serve as the predominant pollinators in Hawai'i forests where as many as 60% of the native trees are bird pollinated. In addition to being important to Hawai'i’s forest ecosystems, these vibrant birds were prominent in traditional native Hawaiian culture. Their feathers were symbols of power and were collected by the thousands to craft intricate capes, headwear, and lei for royalty. Honeycreepers are featured in many native stories, and were often seen as guardian spirits. Sadly, of the many honeycreepers that once inhabited the islands, only 17 species remain, most of them listed as endangered by the IUCN. Deforestation and introduced species, from non-native predators like cats and mongoose, to habitat altering plants like strawberry guava, have been particularly damaging to this group of birds. The most pressing modern threat to honeycreeper survival is the introduced disease avian malaria and its non-native mosquito vector, which is expanding up in elevation as a result of climate change and restricting honeycreepers to the precious little remaining high elevation forest.

### 'I'iwi (*Drepanis coccinea*)
With its crimson plumage and prominent curved bill, the 'I'iwi is one of Hawai'i’s most recognizable birds. Their conspicuous bills are the result of coevolution to match the shape of many native flowers from which they sip nectar. Red was a sacred color for native Hawaiians and 'I'iwi feathers were reserved for use in feather craft for high ranking members of society. Today, 'I'iwi have an estimated population of over 600,000 spread across Hawai'i, Maui Nui, O'ahu, and Kaua'i, but are listed as vulnerable by the IUCN and are federally listed as threatened. While the primary vocalizations made by 'I'iwi are a discordant ‘rusty hinge’ song and a loud, squeaky call, their vocal repertoires are highly diverse, and they will often mimic other birds.

### 'Ākohekohe (*Palmeria dolei*)
The 'Ākohekohe are the largest extant honeycreeper species and are also the most territorial. They have a prominent tufted crest, giving rise to their English name the Crested Honeycreeper, and are the only living honeycreeper with predominantly black feathers. 'Ākohekohe are endemic to Maui Nui, and today are restricted to the remote northeastern slope of Haleakalā volcano, where only about 3,800 individuals are thought to remain. Due to their small population size and restricted range, the IUCN has them listed as critically endangered, and they are federally listed as endangered as well. 'Ākohekohe have a variety of low, raspy, guttural calls and whistle calls.

### Kiwikiu (*Pseudonestor xanthophrys*)
Kiwikiu are one of the rarest of the honeycreepers, with only about 300 individuals remaining on the northeastern slope of Haleakalā volcano on Maui, earning them an IUCN listing of critically endangered as well as a federal listing of endangered. The original Hawaiian name for this bird was lost to history, and they were known as the Maui Parrotbill, aptly chosen for its large bill used to rip into tree branches looking for invertebrates, until they received their new Hawaiian name in 2010. ### 'Akiapōlā'au (*Hemignathus wilsoni*)
'Akiapōlā'au are endemic to Hawai'i Island, and fill a similar ecological niche to woodpeckers. 'Akiapōlā'au use their thick, shorter lower mandible to make holes in the bark, then use their long, curved upper mandible to probe for insects in the tree bark. They are considered endangered both federally and by the IUCN, and less than 1,000 individuals are estimated to remain in high elevation forest habitats on the island. Their slow reproduction makes them particularly vulnerable to extinction. 'Akiapōlā'au have a variety of songs and calls, with a primary song consisting of a fast paced warble often ending in a short trill, complex subsongs that often include snippets of songs from other sympatric bird species, and slurred whistle calls.

### 'Alawī (*Loxops mana*)
Owing to their similarity in color and habitat, the cryptic 'Alawī is often mistaken for another honeycreeper, the Hawai'i 'Amakihi, even to the point that their original Hawaiian name was lost for many years. After extensive research in early Hawaiian literature, they finally regained their Hawaiian name again in 2017. This bird’s English name, the Hawai'i Creeper, is derived from their behavior of creeping along branches with their heads on a swivel, looking under every inch of bark for their insect prey. 'Alawī are listed as endangered both federally and by the IUCN, and less than 14,000 individuals remain in the high elevation forests on Hawai'i Island. Their primary song is a rapid, descending trill, but they also sometimes sing quiet and variable warbling “whisper” songs. 'Alawī calls are short, ascending whistles.

### Hawai'i 'Amakihi (*Chlorodrepanis virens*)
The Hawai'i 'Amakihi is one of the most abundant honeycreeper species, with more than 800,000 individuals in the forests of Hawai'i, Maui, and Moloka'i Islands. 'Amakihi are not federally listed under the Endangered Species Act and the IUCN considers them a species of least concern. Part of their success is due to low-elevation 'amakihi populations developing some level of immunity to avian malaria, giving conservationists hope that other endangered honeycreepers may also be able to adapt to introduced diseases. 'Amakihi have two songs, a short, quick trill, and a soft, complex, warbling “whisper” song, as well as some nasal calls.

## Thrushes
Historically, there were at least five distinct species of thrushes in the Hawaiian archipelago, but arguments have been made for additional subspecies. Hawaiian thrushes predominantly feed on fruit and insects, and serve as some of the most important seed dispersers for native fruit bearing plants. Because of their characteristic drab plumage, thrush feathers were not often used in native Hawaiian feather craft, but their songs are renowned for being some of the most lovely in the forest. Whatsmore, Hawaiian thrushes produce “whisper songs,” quietly muttered versions of their primary songs. While they are not completely immune, Hawaiian thrushes show much higher resistance to avian malaria than most other forest bird species, though that does not mean they have been unaffected by human disturbance. The primary threats to the remaining thrush species are predation, habitat loss, and habitat degradation from introduced invasive plants and animals. Feral pigs are particularly problematic for thrushes, as they root up forest understories where the majority of the fruit that thrushes rely upon grows. 


### Puaiohi (*Myadestes palmeri*)
The Puaiohi, English name Small Kaua'i Thrush, is one of only two remaining thrush species in the Hawaiian Islands, and is the last remaining native seed disperser for fruiting plants on Kaua'i. With less than 500 individuals estimated to be left in the remote ravines of the Alaka'i Swamp, the IUCN recognizes Puaiohi as critically endangered and they are federally listed as endangered. Puaiohi vocal repertoires lack the diversity displayed by other Hawaiian thrushes. Their songs are simple and reedy, consisting of an introductory whistle, a long trill, and sharp concluding whistles. They often whisper this same song, and will also give distinct, raspy hissing calls.

## Seabirds
Seabirds are capable of long over-ocean flight, and as a result a number of seabird species have colonies on the Hawaiian archipelago. Many of these species that colonized the islands have diverged enough from other seabird populations to have Hawaiian subspecies designations. Historically, Polynesean voyagers and fishermen would observe seabirds to predict weather patterns and to guide them back to land or fertile fishing areas, a tactic still used today. Further, their feathers were used in capes and lei, and seabird chicks were a delicacy served to royalty. Seabirds are important drivers of nutrient cycling on islands where they deposit nutrients sourced from the ocean into terrestrial ecosystems, making them more productive. Because many seabird species in Hawai'i are ground nesting, human introduced predators such as rats, cats, and mongoose have been devastating to their populations, and introduced grazers like cows and sheep have degraded their nesting habitat extensively. Seabirds use the moon and stars to navigate back to their nesting sites at night, which makes them victims of an unusual threat, light pollution. Bright lights at night can confuse and disorient seabirds, often leading to collisions with walls, power lines, and other human made objects. Once grounded, seabirds are generally unable to take flight again, making them vulnerable to predation and other harm.

### 'U'au (*Pterodroma sandwichensis*)
The 'U'au is a subspecies of Dark-rumped Petrel which only nests in the Hawaiian Islands, earning them the English name Hawaiian Petrel. They were once abundant on all of the main islands in the Hawaiian archipelago, digging their nesting burrows high up on volcanic slopes. However, today 'U'au colonies are greatly diminished with fewer than an estimated 52,000 individuals remaining nesting on Hawai'i, Maui Nui, and Kaua'i. The subspecies is federally listed as well as listed by the IUCN as endangered. The Hawaiian name for this seabird is derived from the sound of its call (*uuu-A-uu*) which resonates on the mountain sides where they nest. They also produce a softer, more raspy call, and various squeaks.

### 'Ake'ake (*Oceanodroma castro*)
The 'Ake'ake is the Hawaiian breeding population of Band-rumped Storm-Petrel, and is also the rarest, the smallest, and one of the highest nesting of the seabirds that breed in the archipelago. Band-rumped Storm-Petrels are common in other parts of the world, and are a species of least concern to the IUCN, however the 'Ake'ake is federally listed as endangered owing to the small population remaining in the islands. It is estimated less than 400 individuals may still breed somewhere on the islands of Hawai'i, Maui Nui, and Kaua'i, but their burrows are in remote areas that are difficult to access, making surveys challenging. 'Ake'ake calls made at their burrows sound like a guttural purring punctuated by sharp squeaks, while calls made during flight are a rapid, raspy chatter.
