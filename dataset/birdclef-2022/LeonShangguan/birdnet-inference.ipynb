```python
import glob
import pandas as pd
result_files = glob.glob("./*.txt")
```

```python
pred_dicts = {}
for file in result_files:
    data = pd.read_csv(file, sep='\t')
    for time in range(5, 65, 5):
        pred = []
        pre = file.split('.')[1][1:] + '_' + str(time)
        labels = data[(time-5 <= data["Begin Time (s)"]) & (data["Begin Time (s)"] < time)]["Species Code"].values.tolist()
        labels += data[(time-5 <= data["End Time (s)"]) & (data["Begin Time (s)"] < time)]["Species Code"].values.tolist()

        pred = set(labels)
#         print(pred, " ".join(pred))
#         for l in labels:
#             if l not in pred:
#                 pred.append(l)
#         print("*******************************************************************")
#         print(pre, pred)
        pred_dicts[pre] = " ".join(pred)
```

```python
pred_dicts
```

```python
sample_submission = pd.read_csv('../input/birdclef-2022/sample_submission.csv')
sample_submission
```

```python
for i in range(len(sample_submission)):
    sample = sample_submission.row_id[i]
    key = sample.split("_")[0] + "_" + sample.split("_")[1] + "_" + sample.split("_")[3]
    target_bird = sample.split("_")[2]
    print(key, target_bird)
    if key in pred_dicts:
        sample_submission.iat[i, 1] = (target_bird in pred_dicts[key])
```
