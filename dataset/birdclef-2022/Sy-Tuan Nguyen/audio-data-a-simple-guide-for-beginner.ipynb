```python
import os, random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import IPython.display as ipd
import librosa
from scipy.fft import fft, ifft

import warnings
warnings.simplefilter('ignore')
```

# Get a random audio file

```python
AUDIO_DIR = '../input/birdclef-2022/train_audio'
folder = random.choice(os.listdir(AUDIO_DIR))
file   = random.choice(os.listdir(f'{AUDIO_DIR}/{folder}'))
```

```python
y, sr = librosa.load(f'{AUDIO_DIR}/{folder}/{file}')
```

```python
print(y)
```

```python
print(sr)
```

# Play the selected audio file

```python
ipd.Audio(y, rate=sr)
```

# Plot the audio timeseries

```python
plt.plot(y, c=plt.cm.get_cmap('tab10', 10)(np.random.randint(10)), linewidth=1);
```

# Fast Fourier Transform (fft)

```python
# real part
plt.plot(fft(y).real, c=plt.cm.get_cmap('tab10', 10)(np.random.randint(10)), linewidth=1);
```

```python
# imaginary part
plt.plot(fft(y).imag, c=plt.cm.get_cmap('tab10', 10)(np.random.randint(10)), linewidth=1);
```

# Inverse Fourier Transform

```python
plt.plot(ifft(fft(y)), c=plt.cm.get_cmap('tab10', 10)(np.random.randint(10)), linewidth=1);
```
