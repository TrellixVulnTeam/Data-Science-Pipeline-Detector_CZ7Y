<h2 style="font-weight: bold;">1. Indian Peafowl</h2>



<div class ="image"> <img style="float: left;margin:5px 20px 5px 1px" src="https://cdn.download.ams.birds.cornell.edu/api/v1/asset/97326721/1800" width="400" height="500"></div>
<p>Peafowl is a common name for three bird species in the genera Pavo and<br> Afropavo within the tribe Pavonini of the family Phasianidae,<br>the pheasants and their allies.</p>

```python
# function
# name represents the selected bird
#num represents number of audio in the selected bird

def bird(num,name):
    loca = india_df[india_df['common_name'] == name]["filename"].values[num]
    all = link+'/train_audio/'+loca
    #diplay wave
    ar, sr = librosa.load(link+'/train_audio/'+loca, duration=10)
    arr = librosa.stft(ar)
    arr_db = librosa.amplitude_to_db(abs(arr))
    librosa.display.specshow(arr_db, sr=sr, x_axis='s', y_axis='hz')
    plt.colorbar()
    return ipd.Audio(all)

bird(10,"Indian Peafowl")
```
