## 1. 「今この鳥鳴いている？」の２値分類問題である

基本的に test.csv の行ごとに、「この音声ファイル（file_id）のこの時間（end_time-5 -> end_time）にこの鳥（bird）が鳴いているか否か」を当てる２値分類問題のようです。test.csv の各項目の意味についてまとめます：

|項目名|意味|
|---|---|
|row_id|行ID。予測を行う単位となります。|
|file_id|テスト用音声データのファイルID。拡張子oggを忘れずに。|
|bird|「今この鳥鳴いてる？」の「この鳥」のこと。|
|end_time|「今この鳥鳴いてる？」の「今」がどのタイミングかということ。|

ちなみに提出するファイルのサンプル（sample_submit.csv）を見ると、以下のような項目になっています。True or Falseの２値分類になってますね。

|項目名|意味|
|---|---|
|row_id|行ID。予測を行う単位となります。|
|target|birdで指定された鳥が鳴いていればTrueです。そうでなければFalse。|

学習モデルを作るときは、なんか鳥がピヨピヨ鳴いてるみたいな音声データ（ogg形式）が /kaggle/input/train/鳥の名前/XC[数字].ogg というファイル名で大量に与えられているので、それをもとに学習モデルを作ることになります。また、train_metadata.csvには音源が採取された緯度経度のような情報も含まれており、音声コンペだけでなくテーブルコンペ風味も加わっています。

## 3. 評価指標は macro-F1 を用いる

評価指標は macro-F1 であると記載されています。macro-F1 では、対象となる鳥の種類ごとのF1値の平均となります。一旦鳥の種類ごとに集計する手間が必要になる点が異なるわけですね。sklearn.metrics.f1_scoreを使う場合は、きちんと引数に average='macro' と明記しましょう。
