# The centroid of a collection of points is just the componentwise mean

Given a collection of points `{(x1,y1),...,(xn,yn)}`, the mean is just the pair `(1/n)*(∑i xi, ∑i yi)`

We can do this with a dataframe by selecting the `x`, `y`, and `place_id` columns,
grouping by the `place_id` and then calling `mean()` on the result.

```python
df_train[["x","y","place_id"]].groupby("place_id").mean()
```
