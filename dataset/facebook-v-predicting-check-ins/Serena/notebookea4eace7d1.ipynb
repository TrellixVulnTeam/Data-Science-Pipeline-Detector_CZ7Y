```python
df = pd.read_csv("../input/train.csv")
print("done")
```

The check-in locations seem somewhat evenly distributed.

```python
df.sample(frac=.001).plot(kind='hexbin', x='x', y='y', gridsize=25)
```

There are clear peaks and troughs in check-in times

```python
df.sample(frac=.001)["time"].plot(kind="density")
```

For each `place_id`, how spread out are check-in locations? We plotted the ranges for `x` and `y` for each `place_id`.

```python
places = df.groupby('place_id')
dists = places.max() - places.min()
```

```python
stds = places.std()
# IDK why it won't let me plot this...
```

```python
plt.violinplot([dists["x"], dists["y"]])
```
