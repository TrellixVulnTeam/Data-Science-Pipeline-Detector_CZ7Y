```python
import numpy as np
import pandas as pd

train = open('/kaggle/input/hashcode-2021-oqr-extension/hashcode.in').read().split('\n')
D, I, S, V, F =  map(int, train[0].split(' '))
streets = [x.split(' ') for x in train[1:S]]
vehicles = [[x.split(' ')[0], x.split(' ')[1:]] for x in train[S+1:S+1+V]]
vehicleState = {i:{'streetsNum': k[0], 'arr': k[1], 'curIndex': 0, 'state': 'queue', 'timeConsume': 0, 'onRoadSecond' : 0} for i,k in enumerate(vehicles)}
dstreets = {str(x[2]): {'B': int(x[0]), 'E': int(x[1]), 'L': int(x[3]), 'C': 0} for x in streets}

schedule = {}
for i_ in range(len(vehicles)):
    for i in range(int(vehicles[i_][0])):
        if vehicles[i_][1][i] not in dstreets:
            dstreets[vehicles[i_][1][i]] = {'B': dstreets[vehicles[i_][1][i-1]]['E'], 'E': dstreets[vehicles[i_][1][i+1]]['B'], 'L': 10, 'C': 0}
        dstreets[vehicles[i_][1][i]]['C'] += 1
        I_ = str(dstreets[vehicles[i_][1][i]]['E'])
        if I_ not in schedule:
            schedule[I_] = [str(vehicles[i_][1][i])]
        else:
            schedule[I_].append(vehicles[i_][1][i])
            schedule[I_] = list(set(schedule[I_]))

s = open('submission.csv', 'w')
s.write(str(len(schedule)) + '\n')
for k in schedule:
    s.write(str(k) + '\n')
    s.write(str(len(schedule[k])) + '\n')
    ss = [v[1] for v in sorted([[dstreets[k1]['L'] + (1-(dstreets[k1]['C']/10)), k1] for k1 in schedule[k]], reverse=True)]
    for st in ss:
        if dstreets[st]['C'] > 7:
            s.write(str(st) + ' ' + str(2) + '\n') 
        else:
            s.write(str(st) + ' ' + str(1) + '\n') 
s.close()
```
