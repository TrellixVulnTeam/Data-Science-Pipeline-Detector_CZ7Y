```python
code = """
#include<bits/stdc++.h>
#define fast ios::sync_with_stdio(0); cin.tie(0); cout.tie(0)
#define ll long long int
#define pb push_back
#define mk make_pair
#define F first
#define S second
#define all(x) x.begin(), x.end()
#define vi vector<int>
#define vl vector<long>
#define rep(i, a, b) for (int i = a; i < b; i++)


// For Kaggle 
#define input "/kaggle/input/hashcode-2021-oqr-extension/hashcode.in"
#define output "submission.csv"

using namespace std;

class Input
{
public: 
    int Time, Intersection, Edges, Cars, bp;  
    Input()
    {
        Time = 0;
        Intersection = 0; 
        Edges = 0;
        Cars = 0;
        bp = 0;
    }  
};

/** Algo:
    * I am moving on the road and checking
    * end points of road.
    * ex - for car1 :-> 0, 1, then 2;
    * if it's last road then don't count 
    * for that I used condition if last road
    * then continue;
    */
int cal_intersection(map<int, int> cartrip, vector<vector<string>>& carspath, map<string, tuple<int,int,int>>&roads)
{
   int cnt_of_intersection = 0; 
   for (auto x : carspath)
   {
    for (int j = 0; j < x.size(); j++)
    {
        if (j == x.size() - 1)
        {
            continue;
        }
        int u, v, t;
        tie(u, v, t) = roads[x[j]];
        if (cartrip[v] == 0)
        {
            cartrip[v]++;
            cnt_of_intersection++;
        }
    }
   }
   return cnt_of_intersection;
}

int main()
{
   // freopen("input.txt", "r", stdin);
   // freopen("output.txt", "w", stdout);
   
   // For Kaggle use this.
   freopen(input, "r", stdin);
   freopen("submission.csv", "w", stdout);
   
   fast;
   Input obj = Input();

   //First input line.
   cin>>obj.Time>>obj.Intersection>>obj.Edges>>obj.Cars>>obj.bp;

   //Checking of first input line.
   // cout<<obj.Time<<" "<<obj.Intersection<<" "<<obj.Edges<<" "<<obj.Cars<<" "<<obj.bp<<endl;
   
   //Containers for roads and carpath;
   map<string, tuple<int, int, int>> roads;
   vector<vector<string>> carspath;

   //Extra containers which will calculate
   //inwards streets and visited of intersection;
   map<int, vector<string>> inward_edges;
   map<int, int> cartrip;

   //Varible to calculate intersection;
   int cnt_of_intersection = 0;

   //Input of roads.
   for (int i = 0, u, v, t; i < obj.Edges; i++)
   {
       string s;
       cin>>u>>v>>s>>t;
       roads.insert({s, {u, v, t}});
       inward_edges[v].pb(s);
   }

   //Input of each car.
   for (int i = 0, u; i < obj.Cars; i++)
   {
       cin>>u;
       vector<string> temp;
       for (int j = 0; j < u; j++)
       {
          string s;
          cin>>s;
          temp.pb(s);
       }
       carspath.pb(temp);
   }

   // //checking of roads input.
   // for (auto x : roads)
   // {
   //    int u, v, t;
   //    tie(u, v, t) = x.S;
   //    cout<<x.F<<" => "<<u<<" "<<v<<" "<<t<<endl;
   // }
   
   vector<int> vis(obj.Intersection, 0);
   // Calculation and checking for no. of intersection;
   cnt_of_intersection = cal_intersection(cartrip, carspath, roads);
   // cout<<"No. of Intersection: "<<cnt_of_intersection<<endl;
   

   //Output:
   cout<<cnt_of_intersection<<endl;
   for (auto x : carspath)
   {
    for (int j = 0; j < x.size(); j++)
    {
        if (j == x.size() - 1)
        {
            continue;
        }
        int u, v, t;
        tie(u, v, t) = roads[x[j]];
        if (!vis[v])
        {
            vis[v] = 1;
            cout<<v<<endl;
            int no_of_inward = inward_edges[v].size();
            cout<<no_of_inward<<endl;
            for (auto y : inward_edges[v])
            {
                int val = 1 + (rand() % 1);
                cout<<y<<" "<<(val)<<endl;
            }
        }
    }
   }
}"""
```
