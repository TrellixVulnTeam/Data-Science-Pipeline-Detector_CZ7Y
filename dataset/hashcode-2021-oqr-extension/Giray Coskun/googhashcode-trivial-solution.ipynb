```python
## Read File

input_file = open(input_filename, 'r')

## Parse Input

first_line = input_file.readline()
first_line = first_line.split()
first_line = [int(i) for i in first_line]
print(first_line)

Duration = first_line[0]
Intersections = first_line[1]
Streets = first_line[2]
Vehicles = first_line[3]
Score = first_line[4]
```

```python
street_dict = {}

for street in range(Streets):
    line = input_file.readline()
    elements = line.split()
    numbers = [int(i) for i in elements if i.isnumeric()]
    street_dict[elements[2]] = numbers
    

"""
{'rue-de-londres': [2, 0, 1],
 'rue-d-amsterdam': [0, 1, 1],
 'rue-d-athenes': [3, 1, 1],
 'rue-de-rome': [2, 3, 2],
 'rue-de-moscou': [1, 2, 3]}
"""
```

```python
intersection_dict = {}

for key, value in street_dict.items():
    item = (value[0], value[1])
    if item in intersection_dict:
        intersection_dict[item].append(key)
    else:
        intersection_dict[item] = [key]
        
"""
{(2, 0): ['rue-de-londres'],
 (0, 1): ['rue-d-amsterdam'],
 (3, 1): ['rue-d-athenes'],
 (2, 3): ['rue-de-rome'],
 (1, 2): ['rue-de-moscou']}
"""
```

```python
section_dict = {}

for key, value in street_dict.items():
    item = str(value[1])
    if item in section_dict:
        section_dict[item].append(key)
    else:
        section_dict[item] = [key]
        
"""
{'0': ['rue-de-londres'],
 '1': ['rue-d-amsterdam', 'rue-d-athenes'],
 '3': ['rue-de-rome'],
 '2': ['rue-de-moscou']}
"""
```

```python
vehicle_num = []
vehicles_street = []
for vehicle in range(Vehicles):
    line = input_file.readline()
    elements = line.split()
    #print(elements)
    #numbers = [int(i) for i in elements if i.isnumeric()]
    vehicle_num.append(int(elements[0]))
    vehicles_street.append((elements[1:]))

    
"""

vehicle_num = [4,3]

vehicles_street = 

[['rue-de-londres', 'rue-d-amsterdam', 'rue-de-moscou', 'rue-de-rome'],
 ['rue-d-athenes', 'rue-de-moscou', 'rue-de-londres']]
 
"""
```

```python
output_filename = "submission.csv"
output_file = open(output_filename, 'w')
time = 1

for k in range(Intersections):
    if (str(k) in section_dict):
        output_file.write(str(k)+'\n')
        output_file.write(str(len(section_dict[str(k)]))+'\n')
        if (len(section_dict[str(k)]) == 1):
            text = section_dict[str(k)][0]+ ' ' + str(Duration)
            output_file.write(text+'\n')
        else:
            for street in section_dict[str(k)]:
                text = street + ' ' + str(int(time)) 
                output_file.write(text+'\n')
                
output_file.close()
```
