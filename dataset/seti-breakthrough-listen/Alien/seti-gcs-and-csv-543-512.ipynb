```python
from kaggle_datasets import KaggleDatasets
import pandas as pd
```

```python
GCS_DS_PATH09 = KaggleDatasets().get_gcs_path("seti-npy-543512-09")
GCS_DS_PATHaf = KaggleDatasets().get_gcs_path("seti-npy-543512-af")
```

```python
df = pd.read_csv('../input/seti-breakthrough-listen/train_labels.csv')
df['0_f'] = 0
for i in range(df.shape[0]):
    df.loc[i, '0_f'] = df.loc[i, 'id'][0]
```

```python
df['gcs'] = 0
for i in range(31415):
    df.loc[i, 'gcs'] = GCS_DS_PATH09 + '/' + df.loc[i, 'id'] + '.npy'
for i in range(31415, 50165 ):
    df.loc[i, 'gcs'] = GCS_DS_PATHaf + '/' + df.loc[i, 'id'] + '.npy'
```
