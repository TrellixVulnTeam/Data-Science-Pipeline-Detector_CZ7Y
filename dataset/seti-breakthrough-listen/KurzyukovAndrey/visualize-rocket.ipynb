### Load 

```python
cadence = np.load('../input/seti-breakthrough-listen/train/6/6c12bab0aeb4.npy').astype(np.float32)
```

### Utils

```python
def plot_cadence(cadence: np.array) -> None:
    """Visualize/plot 6 subplots."""
    
    assert cadence.shape == (6, 273, 256)
    
    num_imgs = cadence.shape[0]
    fig, ax = plt.subplots(num_imgs, figsize=(10, 40))
    for i in range(num_imgs):
        ax[i].imshow(cadence[i])
    plt.show()
```

### Visualize

```python
plot_cadence(cadence)
```
