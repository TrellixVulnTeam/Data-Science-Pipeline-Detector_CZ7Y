# SETI - What Do Signals Look Like?

```python
import numpy  as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib
import matplotlib.pyplot as plt

import os
import numpy as np
import pandas as pd
from PIL import Image
from matplotlib import pyplot as plt
import seaborn as sns
```

Load data and resolve filenames from ids

```python
def get_train_file_path(image_id): return "../input/seti-breakthrough-listen/train/{}/{}.npy".format(image_id[0], image_id)
def get_test_file_path(image_id):  return "../input/seti-breakthrough-listen/test/{}/{}.npy".format( image_id[0], image_id)

train_df = pd.read_csv('../input/seti-breakthrough-listen/train_labels.csv')
test_df  = pd.read_csv('../input/seti-breakthrough-listen/sample_submission.csv')

train_df['file_path'] = train_df['id'].apply(get_train_file_path)
test_df['file_path']  = test_df['id'].apply(get_test_file_path)

display(train_df.head())
display(test_df.head())
```

Plot signal/no-signal data both as indivual parts and vstacked 

```python
def plot_image_parts(df_row, rows=1, cols=6):
    total = rows * cols 
    plt.figure(figsize=(100/cols, rows*2))
    # plt.subplots_adjudf_rowst(left=None, bottom=None, right=None, top=None, wspace=0, hspace=1)
    image = np.load(df_row['file_path'])    # (6, 273, 256)
    image = image.astype(np.float32)
    for i in range(6):
        plt.subplot(rows, cols, i + 1)
        plt.title(f"{df_row['id']} = {df_row['target']}")
        plt.imshow(image[i])
    plt.show()

def plot_vstack(df, rows=10, cols=2):
    total = rows * cols 
    plt.figure(figsize=(48/cols, rows*2))
    # plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0, hspace=1)
    for i in range(total):
        image = np.load(df.iloc[i]['file_path'])    # (6, 273, 256)
        image = image.astype(np.float32)
        image = np.vstack(image).transpose((1, 0))  # (1638, 256) -> (256, 1638)
#         image = np.hstack(image).transpose((1, 0))  # (1638, 256) -> (256, 1638)
        plt.subplot(rows, cols, i + 1)
        plt.title(f"{df.iloc[i]['id']} = {df.iloc[i]['target']}")
        plt.imshow(image)
    plt.show()
    
    
# Source: https://www.kaggle.com/ihelon/signal-search-exploratory-data-analysis
def plot_cadence(df_row) -> None:
    filename = df_row['file_path']
    label    = df_row['target']
            
    plt.figure(figsize=(8, 5))
    arr = np.load(filename)
    for i in range(6):
        plt.subplot(6, 1, i + 1)
        if i == 0:
            plt.title(f"ID: {os.path.basename(filename)} TARGET: {label}", fontsize=18)
        plt.imshow(arr[i].astype(float), interpolation='nearest', aspect='auto')
        plt.text(5, 100, ["ON", "OFF"][i % 2], bbox={'facecolor': 'white'})
        plt.xticks([])
    plt.show()
```

```python
print('Signal')
plot_image_parts(train_df.query('target == 1').iloc[0])
plot_vstack(train_df.query('target == 1'), cols=1, rows=1)
plot_cadence(train_df.query('target == 1').iloc[0])
```

```python
print('No Signal')
plot_image_parts(train_df.query('target == 0').iloc[0])
plot_vstack(train_df.query('target == 0'), cols=1, rows=1)
plot_cadence(train_df.query('target == 0').iloc[0])
```

Now show a larger number of samples?


```python
print('Signal')
for i, df_row in train_df.query('target == 1').iterrows(): 
    plot_cadence(df_row)
```
