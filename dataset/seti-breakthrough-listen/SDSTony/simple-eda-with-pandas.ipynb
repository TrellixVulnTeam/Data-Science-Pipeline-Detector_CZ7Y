```python
import pandas as pd
import glob
import numpy as np
import matplotlib.pyplot as plt
```

```python
len(glob.glob('../input/seti-breakthrough-listen/train/*')), len(glob.glob('../input/seti-breakthrough-listen/test/*'))
```

There are 16 major folders in `train` and `test` folder. 

```python
def plot_image_num_dist(folder: str):
    major_folders = sorted(glob.glob(f'../input/seti-breakthrough-listen/{folder}/*'))
    dic = {}
    for folder in major_folders:
        dic[f'{folder[-1]}'] = len(glob.glob(folder + '/*'))
    pd.Series(dic).plot(kind='barh')
```

```python
plot_image_num_dist('train')
```

We have about even numbers of images in each major folder. Each folder contains about 3,000 images. 

```python
plot_image_num_dist('test')
```

The distribution of images in `test` folder is somewhat similar to that of the `train` folder. Each folder contains about 2,000 images.
