```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from matplotlib import pyplot as plt
import cv2
import os
```

```python
error_pre = np.load('../input/error-predict/error_file.npy')
plt.figure(figsize=(15, 15))
for i in range(0,30):
    file_name = error_pre[i].tolist().split('listen/')[-1]
    image = np.load(os.path.join('../input/seti-breakthrough-listen',file_name)) 
    image = image.astype(np.float32)
    image = np.vstack(image).transpose((1, 0)) 
    plt.figure(figsize=(15, 15))
    plt.imshow(image)
    plt.show()
    
```
