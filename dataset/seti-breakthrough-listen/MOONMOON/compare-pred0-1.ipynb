```python
import pandas as pd
import numpy as np
from functools import reduce
import matplotlib.pyplot as plt
import matplotlib
import os
matplotlib.rc("font",family='MicroSoft YaHei',weight="bold")
from matplotlib.pyplot import MultipleLocator


df = pd.read_csv('../input/compare-pred/allfold_eval_pred.csv')

pred_0 = df[df['target'] ==0]
print(len(pred_0))
plt.figure(figsize=(16,8))

list0=[]
for i in range(11):
    list0.append(float(i*0.1))
    
x_major_locator=MultipleLocator(0.1)
ax=plt.gca()
ax.xaxis.set_major_locator(x_major_locator)

n, bins, patches = plt.hist(pred_0.preds, bins = list0, color='brown', alpha = 0.8, label = "直方图" ) 

for i in range(len(n)):
    plt.text(bins[i]*1., n[i]*1.01, int(n[i]),fontsize = 16) 


plt.xlabel('TARGET:0  preds',fontsize=25)

plt.ylabel('Number',fontsize=25)
plt.grid(alpha = 0.5)
plt.savefig('./train0_pred.png') 
```

```python
pred_1 = df[df['target'] ==1]
print(len(pred_1))
plt.figure(figsize=(16,8))

list1=[]
for i in range(21):
    list1.append(float(i*0.05))
    
x_major_locator=MultipleLocator(0.05)
ax=plt.gca()
ax.xaxis.set_major_locator(x_major_locator)

n, bins, patches = plt.hist(pred_1.preds, bins = list1, color='brown', alpha = 0.8, label = "直方图" )

for i in range(len(n)):
    plt.text(bins[i]*1., n[i]*1.01, int(n[i]),fontsize = 16) 

plt.grid(alpha = 0.5)

plt.xlabel('TARGET:1 preds',fontsize=25)

plt.ylabel('Number',fontsize=25)
plt.grid(alpha = 0.5)
plt.savefig('./train1_pred.png') 
```

```python
pred_0_bad = pred_0[pred_0['preds'] >0.9]
pred_0_bad
```

```python
def get_train_filename_by_id(_id: str) -> str:
    return f"../input/seti-breakthrough-listen/train/{_id[0]}/{_id}.npy"

def show_cadence(filename: str, label: int,pred: float,name ) -> None:
    fig, axes = plt.subplots(6, 1, figsize = (16, 10))
    ax = axes.ravel()
    arr = np.load(filename)
    for i in range(6):
        
        ax[i].imshow(arr[i].astype(float), interpolation='nearest', aspect='auto')
        ax[i].text(5, 100, ["ON", "OFF"][i % 2], bbox={'facecolor': 'white'})
        if i != 5:
            ax[i].set_xticks([])
            
    fig.text(0.5, -0.02, 'Frequency Range', ha='center', fontsize=18)
    fig.text(-0.02, 0.5, 'Seconds', va='center', rotation='vertical', fontsize=18)

    plt.suptitle(f"ID: {os.path.basename(filename)} TARGET: {label} Pred: {pred}", fontsize=18)
    fig.tight_layout()

    plt.show()
    
    '''save'''
#     if label ==0:
#         plt.savefig('./'+name+'.png') 
#     elif label ==1:
#         plt.savefig('./'+name+'.png') 
```

```python
for i in range(len(pred_0_bad)):
    negative_target = pred_0_bad.iloc[i].id
    negative_pred = pred_0_bad.iloc[i].preds
    show_cadence(get_train_filename_by_id(negative_target), 0,negative_pred,negative_target)
```

```python
for i in range(len(pred_1_bad)):
    if i>10:
        break
    positive_target = pred_1_bad.iloc[i].id
    positive_pred = pred_1_bad.iloc[i].preds
    show_cadence(get_train_filename_by_id(positive_target), 1,positive_pred,positive_target)
```
