```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

### In this notebook, I have given similar images between the test and the train. 5 different example with different degrees of similarity. 
##### There are many more similar images in the train and test.

```python
train_labels = pd.read_csv('../input/seti-breakthrough-listen/train_labels.csv')
test_labels = pd.read_csv('../input/seti-breakthrough-listen/sample_submission.csv')
```

```python
def get_train_filename_by_id(_id: str) -> str:
    return f"../input/seti-breakthrough-listen/train/{_id[0]}/{_id}.npy"

def get_test_filename_by_id(_id: str) -> str:
    return f"../input/seti-breakthrough-listen/test/{_id[0]}/{_id}.npy"

def visualization(id_1: str, id_2: str) -> None:
    img_1 = np.load(get_test_filename_by_id(id_1))[::2]
    img_2 = np.load(get_train_filename_by_id(id_2))[::2]
    img_1 = np.vstack(img_1).transpose((1, 0))
    img_2 = np.vstack(img_2).transpose((1, 0))
    fig, ax = plt.subplots(1,2,figsize=(10,10))
    ax[0].imshow(img_1.astype(float), interpolation='nearest', aspect='auto')
    ax[0].set_title('test')
    ax[1].imshow(img_2.astype(float), interpolation='nearest', aspect='auto')
    ax[1].set_title('train')
    plt.show()
```

```python
### example 1
visualization('000bf832cae9ff1','ff6c246c4bc8425')
```
