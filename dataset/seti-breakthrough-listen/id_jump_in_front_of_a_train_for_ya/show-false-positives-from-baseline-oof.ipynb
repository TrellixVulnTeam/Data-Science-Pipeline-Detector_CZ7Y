```python
import os

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_oof = pd.read_csv('../input/seti-baseline-oof/oof_df.csv')
```

```python
clf_thres = 0.5

df_oof_tp = df_oof.query('target == 1 & preds >= @clf_thres')
df_oof_fp = df_oof.query('target == 0 & preds >= @clf_thres')
df_oof_fn = df_oof.query('target == 1 & preds < @clf_thres')

display(df_oof_tp.shape)
display(df_oof_fp.shape)
display(df_oof_fn.shape)
```

```python
def show_cadence(filename: str, label: int, preds) -> None:
    plt.figure(figsize=(16, 10))
    arr = np.load(filename)
    for i in range(6):
        plt.subplot(6, 1, i + 1)
        if i == 0:
            plt.title(f"ID: {os.path.basename(filename)} TARGET: {label} PRED: {round(preds, 5)}", fontsize=18)
        plt.imshow(arr[i].astype(float), interpolation='nearest', aspect='auto')
        plt.text(5, 100, ["ON", "OFF"][i % 2], bbox={'facecolor': 'white'})
        plt.xticks([])
    plt.show()


def get_train_filename_by_id(_id: str) -> str:
    return f"../input/seti-breakthrough-listen/train/{_id[0]}/{_id}.npy"


def show_cadence_from_df(df, n=1):
    df_tmp = df.sample(n)
    for ind, row in df_tmp.iterrows():
        show_cadence(get_train_filename_by_id(row["id"]), row["target"], row['preds'])
```

```python
show_cadence_from_df(df_oof_tp, n=3)
```

```python
show_cadence_from_df(df_oof_fp, n=3)
```

```python
show_cadence_from_df(df_oof_fn, n=3)
```
