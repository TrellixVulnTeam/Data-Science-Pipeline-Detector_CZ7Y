```python
from functools import reduce
```

```python
df = reduce(lambda  left,right: pd.merge(left,right,on=['id'],
                                            how='outer'), [df_1, df_2, df_3, df_4, df_5])
```

```python
df.head()
```

```python
df.columns = ['id', 'c1', 'c2', 'c3', 'c4', 'c5']
```

```python
df['target'] = (.6*df['c1'] + .1*df['c2'] +.1*df['c3'] + .1*df['c4']+ .1*df['c5'])
```
