```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
import matplotlib.pyplot as plt
import seaborn as sns
from glob import glob
```

```python
main_path='/kaggle/input/seti-breakthrough-listen/'
train_npy_files=os.path.join(main_path,'train')
test_npy_files=os.path.join(main_path,'test')
train_df=pd.read_csv(os.path.join(main_path,'train_labels.csv'))
sample_sub_df=pd.read_csv(os.path.join(main_path,'sample_submission.csv'))
```

```python
print('Shape of train dataframe: ',train_df.shape)
train_df.head(10)
```

```python
#Let's take a look at train npy file
ones=train_df['id'][train_df['target']==1].values[:3] #Get file ids of positive target
zeros=train_df['id'][train_df['target']==0].values[:3] #Get file ids of negative target
plt_files=np.concatenate((ones,zeros))

r,c=2,3
fig=plt.figure(figsize=(16,8))
for i in range(1,r*c+1):
    tmp_f=glob(os.path.join(main_path,'train','*',plt_files[i-1]+'.npy'))[0]
    tmp_f=np.load(tmp_f)
    fig.add_subplot(r,c,i)
    plt.specgram(tmp_f.reshape(-1));
    if i-1<=2:
        plt.title('Target: 1')
    else:
        plt.title('Target: 0')
```

```python
#Plot target distribution
pos_cls=np.sum(train_df['target'])
print('Positive Percentage: {:.2f}%'.format((pos_cls/train_df.shape[0])*100))
sns.displot(train_df['target'])

#Data is highly unbalanced 
```
