```python
train=pd.read_csv('/kaggle/input/tabular-playground-series-mar-2022/train.csv')
```

```python
train
```

```python
test=pd.read_csv('/kaggle/input/tabular-playground-series-mar-2022/test.csv')
train=train.drop(['row_id'],axis=1)
test=test.drop(['row_id'],axis=1)
```

```python
train['time'] = pd.to_datetime(train['time'])
test['time'] = pd.to_datetime(train['time'])
```

```python
reg = setup(data = train,
            target = 'congestion',
            normalize = True, #normalisation helps some algorithms
#             data_split_shuffle = False, #so that we do not use "future" observations to predict "past" observations
            use_gpu = True,
            silent = True,
            n_jobs = -1)
```

```python
compare_models(sort='MAE',exclude=['et'])
```

```python
dt=create_model('dt')
```

```python
catboost=create_model('catboost')
```

```python
xgboost=create_model('xgboost')
```

```python
blend_specific = blend_models(estimator_list = [dt,catboost,xgboost])
```

```python
predictions=predict_model(blend_specific,data=test)
```

```python
ef=pd.read_csv('/kaggle/input/tabular-playground-series-mar-2022/test.csv')
output= pd.DataFrame ({'row_id': ef.row_id, 'congestion': predictions.Label})
output.to_csv('submission.csv',index=False)
```
