```python
#We import some libraries to handle the data and some visualization with matplotlib
import matplotlib.pyplot as plt
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)



import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

```

```python
# we import it.
file_path = '/kaggle/input/tabular-playground-series-mar-2022/train.csv'
df = pd.read_csv(file_path)
df.index = df['row_id']
df = df.drop('row_id', axis = 1)
df.head(20)
```

```python
#let's check all data types.
df.info()
```

we are changing to datetime format. Not really insterested in the year right now, so we can format the time so we can get the hours and minutes.

```python
df['time'] = pd.to_datetime(df['time'])
df['time'] = df['time'].dt.strftime('%H:%M')
```

Let's check out some distributions of the data and some of the most recurrent values.

```python
df.direction.value_counts().plot(kind = 'bar', figsize =(10,7))
```

we are catching how many values each direction got to see if our data balance.
We can see the most observations of congestions is in the 'East-Bound' so it'd be interesting to see how directions are correlated with congestions. and the time will be a crucial part because we hope there are more congestion at some day time period or night time periods. 

```python
df[df.direction == 'EB']['congestion'].value_counts().sort_index().plot(kind = 'bar', figsize=(14,9))
```

```python
df[df.direction == 'EB']['congestion'].value_counts()
```

```python
df_por_cada_20_minutes = df.groupby('time')['congestion'].mean()
df_por_cada_20_minutes.plot(kind = 'bar', color = 'b', figsize = (15,10) )
```

I wanted to check the mean congestion time by 20 minutes, although I wanted to try to put into hours and a more visual (friendly) graph.

```python
df_cong = df[df.congestion >= 60]
```

```python
f"The percentage of congestion if congestion >= 60 is :{round(len(df_cong) / len(df) * 100,2)} % "
```

```python
df_cong.groupby('time')['congestion'].agg(np.mean).plot(figsize = (14,8), color = 'y', ylabel = 'Mean Congestion', xlabel = 'Time')
```
