# Introduction

I wanted to do the simplest solution which is to simply apply the median. I did this in another notebook https://www.kaggle.com/taranmarley/fe-eda-prediction/ but I wanted to seperate out this simple baseline into its own notebook.

In real world problem it is always worth establishing simple baselines like this in my experience. As they are the best way to look at the real world predictive power of your model. Also simple operations like adding one to the median or blending simple things, can sometimes give an interesting insight into the data.

```python
# Load Data
df = pd.read_csv("../input/tabular-playground-series-mar-2022/train.csv", infer_datetime_format=True)
test_df = pd.read_csv("../input/tabular-playground-series-mar-2022/test.csv")
```

**Create Location**

```python
df["location"] = df["x"].astype(str) + df["y"].astype(str) + df["direction"]
test_df["location"] = df["x"].astype(str) + df["y"].astype(str) + df["direction"]
```

**Parse as Date**

```python
df['time'] = pd.to_datetime(df['time'])
test_df['time'] = pd.to_datetime(test_df['time'])
```

**Create Date Information**

```python
df['rectified_hour'] = df['time'].dt.hour + (df['time'].dt.minute / 60)
test_df['rectified_hour'] = test_df['time'].dt.hour + (test_df['time'].dt.minute / 60)
```

**Create Location Time Info**

```python
df["location_hour"] = df['rectified_hour'].astype(str) + "_" + df['location'].astype(str)
test_df["location_hour"] = test_df['rectified_hour'].astype(str) + "_" + test_df['location'].astype(str)
```

**Average by time and location**

```python
location_hour_df = df.iloc[9360:,:][['location_hour','congestion']].groupby(['location_hour']).median()
location_hour_dict = {}
for idx, row in location_hour_df.iterrows():
    location_hour_dict[idx] = row[0]
df["location_hour_median"] = df["location_hour"].map(location_hour_dict)
test_df["location_hour_median"] = test_df["location_hour"].map(location_hour_dict)
```

# Create Submission

```python
sample_sub = pd.read_csv("../input/tabular-playground-series-mar-2022/sample_submission.csv")
sample_sub["congestion"] = np.round(test_df["location_hour_median"],0) 
sample_sub.to_csv("submission.csv", index=False)
```
