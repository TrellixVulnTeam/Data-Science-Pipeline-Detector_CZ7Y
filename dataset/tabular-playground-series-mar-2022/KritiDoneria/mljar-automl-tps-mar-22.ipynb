```python
import numpy as np
import pandas as pd
from supervised.automl import AutoML # MLJAR AutoML
```

```python
train = pd.read_csv("../input/tabular-playground-series-mar-2022/train.csv")
test = pd.read_csv("../input/tabular-playground-series-mar-2022/test.csv")
```

```python
x_cols = train.columns[1:-1].tolist()
y_col = train.columns[-1]
```

```python
automl = AutoML(
    mode="Compete", 
    total_time_limit=3600*2,# 6 hours 
    model_time_limit=30*60,
    start_random_models=10,
    hill_climbing_steps=3,
    eval_metric='mae',
    stack_models=False,
    golden_features=True
    
)
automl.fit(train[x_cols], train[y_col])
```

```python
preds = automl.predict(test)
```

```python
sub = pd.DataFrame({"row_id": test.row_id, "congestion": preds})
sub.to_csv("./mljar_tspmar22.csv", index=False)
```
