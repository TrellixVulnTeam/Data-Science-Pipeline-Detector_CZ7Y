```python
train = train[(train.hour >= 12) & (train.weekday == 0)]
```

```python
x_data = train.drop(['time', 'congestion'], axis=1)
y_data = train.congestion

x_test = test.drop('time', axis=1)
```

```python
data = pd.concat([x_data, x_test])
```

```python
data.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder()
data = encoder.fit_transform(data).toarray()
```

```python
x_data = data[:train.shape[0]]
x_test = data[train.shape[0]:]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_val, y_train, y_val = train_test_split(x_data, y_data)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

from torch.utils.data import TensorDataset
from torch.utils.data import DataLoader

import time
```

```python
x_train.shape
```

```python
x_train = torch.tensor(x_train, dtype=torch.float)
y_train = torch.tensor(y_train.values, dtype=torch.float)

x_val = torch.tensor(x_val, dtype=torch.float)
y_val = torch.tensor(y_val.values, dtype=torch.float)

x_test = torch.tensor(x_test, dtype=torch.float)
```

```python
y_train = y_train.view(-1, 1)
y_val = y_val.view(-1, 1)
```

```python
train_dataset = TensorDataset(x_train, y_train)
train_dataloader = DataLoader(train_dataset, batch_size=256, shuffle=True)
```

```python
net = nn.Sequential(nn.Linear(193, 500), nn.BatchNorm1d(500), nn.ReLU(),
                    nn.Linear(500, 300), nn.BatchNorm1d(300), nn.ReLU(),
                    nn.Linear(300, 200), nn.BatchNorm1d(200), nn.ReLU(),
                    nn.Linear(200, 100), nn.BatchNorm1d(100), nn.ReLU(),
                    nn.Linear(100, 50), nn.BatchNorm1d(50), nn.ReLU(), 
                    nn.Linear(50, 1))
```

```python
def init_weight(layer):
    if type(layer) == nn.Linear:
        nn.init.xavier_normal_(layer.weight.data)
```

```python
net.apply(init_weight)
```

```python
criterion = nn.L1Loss()
optimizer = torch.optim.SGD(params=net.parameters(), lr=1e-3, momentum=0.9)
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'

print(f'device {device}')
x_val = x_val.to(device)
y_val = y_val.to(device)

x_test = x_test.to(device)
net = net.to(device)
```

```python
for epoch in range(500):
    loss_all = 0
    data_len =  0
    
    for x, target in train_dataloader:
        x = x.to(device)
        target = target.to(device)
        
        predict = net(x)
        loss = criterion(predict, target)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        loss_all += loss
        data_len += 1
    
    if epoch % 50 == 0:
        print(time.strftime('%Y-%m-%d %H:%M:%S',time.localtime(time.time())), f'epoch {epoch} loss {loss_all / data_len}')
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
with torch.no_grad():
    y_pred = net(x_val)

y_pred = torch.round(y_pred)
val_mae = mean_absolute_error(y_val.cpu(), y_pred.cpu().numpy())
print(val_mae)
```
