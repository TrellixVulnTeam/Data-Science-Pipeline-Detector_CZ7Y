# Don't forget to round your predictions in this competition!

Although most regression algorithms output float values, the targets in this competition are integers. In this situation we can usually improve the score by clipping and rounding the regression output (see discussion post [Why rounding improves the score](https://www.kaggle.com/c/tabular-playground-series-jan-2022/discussion/301249)). 

We'll apply this principle to the output of @mukaseevru's notebook [TPS_Mar_22 FE+LAMA](https://www.kaggle.com/mukaseevru/tps-mar-22-fe-lama-lightautoml).

```python
import pandas as pd
sub = pd.read_csv('../input/tps-mar-22-fe-lama-lightautoml/lightautoml_tabularautoml.csv')
sub['congestion'] = sub['congestion'].round().astype(int)
assert (sub['congestion'] >= 0).all()
assert (sub['congestion'] <= 100).all()
sub.to_csv('submission_rounded.csv', index=False)
sub
```
