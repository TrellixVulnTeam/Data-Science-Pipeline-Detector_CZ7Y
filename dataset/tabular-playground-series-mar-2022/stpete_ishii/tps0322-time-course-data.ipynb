```python
import numpy as np
import pandas as pd
import os
import matplotlib.dates as dates
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
train=pd.read_csv('../input/tabular-playground-series-mar-2022/train.csv')
display(train)
test=pd.read_csv('../input/tabular-playground-series-mar-2022/test.csv')
display(test)
```

```python
print(train['direction'].unique().tolist())
print(train['x'].unique().tolist())
print(train['y'].unique().tolist())
print(train['time'].nunique())
print(train['time'].min(),train['time'].max())
```

```python
print(test['direction'].unique().tolist())
print(test['x'].unique().tolist())
print(test['y'].unique().tolist())
print(test['time'].nunique())
print(test['time'].min(),test['time'].max())
```

```python
train['x_y_direction']=train['x'].astype(str)+'_'+train['y'].astype(str)+'_'+train['direction']
print(train['x_y_direction'].nunique())
```

```python
display(train)
data1=train.copy()
```

```python
x_y_directions=train['x_y_direction'].unique().tolist()
```

```python
fig=make_subplots(specs=[[{"secondary_y":False}]])
for item in x_y_directions[0:3]:
    fig.add_trace(go.Scatter(x=data1['time'][data1['x_y_direction']==item],y=data1['congestion'][data1['x_y_direction']==item],name=item),secondary_y=False,)
fig.update_layout(autosize=False,width=700,height=500,title_text="")
fig.update_xaxes(title_text="Date")
fig.update_yaxes(title_text="Congestion",secondary_y=False)
fig.show()
```

```python
fig=make_subplots(specs=[[{"secondary_y":False}]])
for item in x_y_directions[20:23]:
    fig.add_trace(go.Scatter(x=data1['time'][data1['x_y_direction']==item],y=data1['congestion'][data1['x_y_direction']==item],name=item),secondary_y=False,)
fig.update_layout(autosize=False,width=700,height=500,title_text="")
fig.update_xaxes(title_text="Date")
fig.update_yaxes(title_text="Congestion",secondary_y=False)
fig.show()
```

```python
fig=make_subplots(specs=[[{"secondary_y":False}]])
for item in x_y_directions:
    fig.add_trace(go.Scatter(x=data1['time'][data1['x_y_direction']==item],y=data1['congestion'][data1['x_y_direction']==item].rolling(window=1000).mean(),name=item),secondary_y=False,)
fig.update_layout(autosize=False,width=700,height=500,title_text="")
fig.update_xaxes(title_text="Date")
fig.update_yaxes(title_text="Congestion",secondary_y=False)
fig.show()
```

```python
data2=data1[['x_y_direction','congestion']]
data2=data2.groupby('x_y_direction',as_index=False).mean()
data3=data2.sort_values('congestion',ascending=False)
data3
```

```python
fig = px.bar(data3[:30], x='x_y_direction', y='congestion',title="Congestion mean ranking by x_y_direction")
fig.show()
```
