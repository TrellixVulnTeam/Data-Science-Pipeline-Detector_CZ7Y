## <p style="font-family:newtimeroman; font-size:120%; text-align:center">1. Introduction</p> <a href= '#Introduction'></a>

Before approaching and fitting any model to the dataset / problems, the first thing we do is Exploratory Data Analysis (EDA) on the data. In EDA, we will get basic insights of the data. For this EDA, pandas has been used widely.

What do we do using Pandas?
* df.head() method - view the top 5 rows of the dataset.

* df.tail() method - view the bottom 5 rows of the dataset.

* df.info() method - view concise summary of dataset.

* df.describe() method - view statistical properties of dataset.

* df.shape - gives the dimensions of the dataset.

* df.dtypes - gives the data types of the columns.

* df.columns - view the column names of the dataset.

After completing these basic EDAs, finding the correlation (by using any correlation heatmaps), proceeding further with discriptive summaries and plotting many histograms to find the distribution of the data are done using matplotlib (subplots, countplots, pie, etc.,)

Now, is there any way to get all these infos quick but still get a detailed insights on the data?

**Yes, We can do it. By using the pandas profiling.**

<a id='2'></a>
## <p style="font-family:newtimeroman; font-size:120%; text-align:center">2. Importing Libraries</p> <a href= '#Importing Libraries'></a>

Pandas profiling is an open source Python module with which we can quickly do an exploratory data analysis with just a few lines of code. Instead of just giving us a single output, pandas-profiling tool provides a broadly structured HTML file containing all the relevant information that a typical EDA of all the basic commands and attributes provide.

```python
import pandas as pd # data processing, CSV file I/O 
import pandas_profiling as pp #EDA
```

<a id='3'></a>
## <p style="font-family:newtimeroman; font-size:120%; text-align:center">3. Data Visualization</p> <a href= '#Data Visualization'></a>

Now, let's take a look at EDA performed using the pandas profiling.

```python
train = pd.read_csv("../input/tabular-playground-series-mar-2022/train.csv")
test = pd.read_csv("../input/tabular-playground-series-mar-2022/test.csv")
```

<a id='3.1'></a>
## <p style="font-family:newtimeroman; font-size:90%; text-align:center">3.1 EDA for train data</p> <a href= '#EDA for train set'></a>

```python
pp.ProfileReport(train)
```

<a id='3.2'></a>
## <p style="font-family:newtimeroman; font-size:90%; text-align:center">3.2 EDA for test data</p> <a href= '#EDA for test set'></a>

```python
pp.ProfileReport(test)
```

Storing the report to an HTML file is also possible using pandas profiling!

```python
prof = pp.ProfileReport(test)
prof.to_file(output_file='test.html')
```

If a simple report is prefered rather than this descriptive report, Pandas profiling also provides a simple report.

```python
pp.ProfileReport(train, minimal=True)
```

```python
mini_prof = pp.ProfileReport(train, minimal=True)
mini_prof.to_file(output_file='output.html')
```
