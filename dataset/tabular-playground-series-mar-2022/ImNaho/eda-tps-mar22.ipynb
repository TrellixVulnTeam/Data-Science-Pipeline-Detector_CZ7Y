**check count each congestion**

```python
plt.figure(figsize=(25,5))
sns.barplot(x=train['congestion'],y=train['congestion'].value_counts())
plt.ylabel("count")
```

```python
for df in [train, test]:
    df['date'] = pd.to_datetime(df.time)
    df['year'] = df['date'].dt.year
    df['month'] = df['date'].dt.month
    df['week'] = df['date'].dt.day_name()
    df['day'] = df['date'].dt.day
    df["timestamp"] = df['date'].dt.strftime('%H:%M')
    df["hour"] = df['date'].dt.strftime('%H')
    df["minutes"] = df['date'].dt.strftime('%M')
    df["numoftime"] = df["hour"].astype(int)+df["minutes"].astype(int)/60
```
