```python
def details(df):
    sum_null_values = df.isnull().sum()
    percent_null_values = 100* (sum_null_values/len(df))
    data_type = df.dtypes
    unique_values = df.nunique()

    table = pd.concat([sum_null_values,percent_null_values,data_type,unique_values], axis=1)
    table_col = table.rename(columns = {0 : 'Missing Values', 1 : '% of Total Missing Values', 2 : 'Data_Type', 3: 'Unique values'})
    return table_col
```

```python
dt = DecisionTreeRegressor(max_depth=200, min_samples_leaf=5, min_samples_split=50, random_state=23)
rf = RandomForestRegressor(oob_score= True, n_estimators= 100, min_samples_split= 10, min_samples_leaf= 50, max_samples= 0.8, max_depth= 100, bootstrap= True)
lgbm = lightgbm.LGBMRegressor(max_depth=100, min_child_samples=10, n_estimators=200, objective='regression', random_state=23, subsample=0.7)
```

```python
stack = StackingCVRegressor(regressors=(dt, rf, lgbm),
                            meta_regressor=rf, cv=5,
                            use_features_in_secondary=True,
                            store_train_meta_features=True,
                            shuffle=False,
                            random_state=23)

stack.fit(X, y)

    # Train set prediction
y_train_pred = stack.predict(X)
r2_train = r2_score(y, y_train_pred)

#Mean Absolute error - train
mae_train = mean_absolute_error(y, y_train_pred)

# Test set prediction    
y_test_pred = stack.predict(X_test)
# r2_test = r2_score(y, y_test_pred)

#Mean Absolute error - test
# mae_val = mean_absolute_error(y, y_test_pred)

print('R-Squared_train')
print('='*60)
print(r2_train,"\n")   
print('Mean absolute error - train')
print('='*60)
print(mae_train,"\n") 
```
