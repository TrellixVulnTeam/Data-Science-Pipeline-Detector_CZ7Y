```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train=pd.read_csv("../input/tabular-playground-series-mar-2022/train.csv",index_col=0)
train
```

# Check missing values

```python
train.isnull().sum()
```

# Visualization of the coordinate of the roadway

```python
plt.figure(figsize=(6,6))
plt.scatter(train["x"],train["y"])
plt.xlabel("east-west midpoint coordinate")
plt.ylabel("north-south midpoint coordinate")
```

# Distribution of congestion

```python
plt.figure(figsize=(6,6))
plt.hist(train["congestion"],bins=20)
plt.xlabel("congestion")
plt.ylabel("Frequency")
print("Mean:"+str(np.mean(train["congestion"])))
print("Variance:"+str(np.var(train["congestion"])))
```

classify by direction

```python
print("Number of directions:")
print(train["direction"].value_counts())
```

```python
fig,axes=plt.subplots(2,4,figsize=(10,10))
for num,d in enumerate(train["direction"].unique()):
    if num<=3:
        i=0
    else:
        i=1
    j=num-i*4
    x=train["congestion"][train["direction"]==d]
    axes[i][j].hist(x,bins=int(1+np.log2(len(x))))
    axes[i][j].set_xlabel("congestion (direction:"+d+")")
    axes[i][j].set_ylabel("frequency")
plt.tight_layout()
```

# Time-series plot
Time series plots and correlations are drawn using data with (x, y) = (0,0) and direction "EB".

```python
import matplotlib.dates as mdates
import statsmodels.api as sm
train_EB=train[train["direction"]=="EB"]
data=train_EB[(train_EB["x"]==0)&(train_EB["y"]==0)]
data["time"]=pd.to_datetime(data["time"])
fig, ax = plt.subplots(figsize=(30,5))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%m/%d\n%H:%M'))
plt.plot(data["time"][:1000], data['congestion'][:1000])
```

```python
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(data["congestion"],lags=1000, ax=ax1)
ax1.set_ylim(-0.1,0.2)
```
