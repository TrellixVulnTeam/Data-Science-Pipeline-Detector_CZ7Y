```python
import datetime as dt
train.time=pd.to_datetime(train.time)
train["year"]=train["time"].dt.year
train["month"]=train["time"].dt.month
train["week"]=train["time"].dt.isocalendar().week
train["minute"]=train["time"].dt.minute
train["hour"]=train["time"].dt.hour
train["day"]=train["time"].dt.dayofyear
train["weekday"]=train["time"].dt.dayofweek
train["weekend"]=(train["time"].dt.dayofweek>=5).astype("int")
```

```python
import math
from math import sin,cos,pi
import statistics
```

```python
sin_vals={
    'EB':sin(0),
    'NE':sin(1*(pi/4)),
    'NB':sin(2*(pi/4)),
    'NW':sin(3*(pi/4)),
    'WB':sin(4*(pi/4)),
    'SW':sin(5*(pi/4)),
    'SB':sin(6*(pi/4)),
    'SE':sin(7*(pi/4)),
}
```

```python
cos_vals={
    'EB':cos(0),
    'NE':cos(1*(pi/4)),
    'NB':cos(2*(pi/4)),
    'NW':cos(3*(pi/4)),
    'WB':cos(4*(pi/4)),
    'SW':cos(5*(pi/4)),
    'SB':cos(6*(pi/4)),
    'SE':cos(7*(pi/4)),
}
```

```python
train["sin"]=train["direction"].map(sin_vals)
train["cos"]=train["direction"].map(cos_vals)
```
