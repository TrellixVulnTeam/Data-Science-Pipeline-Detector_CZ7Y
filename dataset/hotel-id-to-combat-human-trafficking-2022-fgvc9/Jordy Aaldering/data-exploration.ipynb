# Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

from PIL import Image
```

```python
DATA_FOLDER  = '/kaggle/input/creation-dataset-512x512/'
TRAIN_FOLDER = DATA_FOLDER + 'train_images/'
```

## Hotel Chain Distribution

```python
data_df = pd.read_csv(DATA_FOLDER + 'train_df.csv')
data_df.head()
```

```python
data_df.describe()
```

```python
chains = data_df.groupby('hotel_id').size()
chains.sort_values(inplace=True)
print('We have a total of', chains.count(), 'hotel chains, and there is an average of', round(chains.mean()), 'pictures per chain, but as it is a skewed representation, we should consider the median which is', chains.median())
```

## Image Size Distribution

```python
sizes = []
for dirname, _, filenames in os.walk(TRAIN_FOLDER):
    for filename in filenames: 
        image_size = Image.open(os.path.join(dirname, filename)).size
        sizes.append(image_size)
```

```python
orig_sizes = []
for dirname, _, filenames in os.walk('/kaggle/input/hotel-id-to-combat-human-trafficking-2022-fgvc9/train_images'):
    for filename in filenames: 
        image_size = Image.open(os.path.join(dirname, filename)).size
        orig_sizes.append(image_size)
```

```python
total_small = sum([1 for x, y in sizes if x < 512 or y < 512])
print(total_small, total_small / len(sizes) * 100)

x, y = zip(*sizes)
plt.scatter(x, y)
```

```python
total_small = sum([1 for x, y in orig_sizes if x < 512 or y < 512])
print(total_small, total_small / len(orig_sizes) * 100)

x, y = zip(*orig_sizes)
plt.scatter(x, y)
```
