```python
import PIL
from typing import Callable

def resize_image(image: PIL.Image, width: int, height: int):
    w_and_h = (width, height)
    resized_image =image.thumbnail(w_and_h)
    return resized_image

#Takes a list, and modifies it so that all images are resized to the specified size
def resize_all_images(hotels: dict, width: int, height: int):
    for hotel in hotels:
        images = hotels.get(hotel, None)
        for i in range(len(images)):
            images[i] = resize_image(images[i], width, height)
            
def filter_hotels_immutable(hotels: dict, function: Callable[[tuple],bool]):
    result = dict(filter(function, hotels.items()))
    return result


def map_all_images(hotels: dict, imageMapFunction: Callable[[any], any]):
    for hotel in hotels:
        images = hotels.get(hotel, None)
        images_mapped = map(imageMapFunction, images)
        hotels.update(hotel, images_mapped)
        
def map_images_filter_hotels(hotels: dict, hotelFilter: Callable[[tuple],bool], imageMapFunction: Callable[[any],any] ):
    filtered_hotels = filter_hotels_immutable(hotels, hotelFilter)
    for hotel in filtered_hotels:
        images = hotels.get(hotel, None)
        images_mapped = map(imageMapFunction, images)
        hotels.update(hotel, images_mapped)
        
def map_images_filter_hotels(hotels: dict, hotelFilter: Callable[[tuple],bool], imageMapFunction: Callable[[any],any] ):
    filtered_hotels = filter_hotels_immutable(hotels, hotelFilter)
    for hotel in filtered_hotels:
        images = hotels.get(hotel, None)
        images_mapped = map(imageMapFunction, images)
        hotels.update(hotel, images_mapped)


    
```
