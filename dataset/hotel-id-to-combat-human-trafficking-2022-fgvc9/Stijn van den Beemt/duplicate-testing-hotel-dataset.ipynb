```python
from PIL import Image as pil_image
import matplotlib.pyplot as plt
```

```python
import hashlib
from tqdm import tqdm
DATA_FOLDER = "../input/hotelid-2022-train-images-256x256/"
IMAGE_FOLDER = DATA_FOLDER + "images/"
OUTPUT_FOLDER = ""

train_df = pd.read_csv(os.path.join(DATA_FOLDER, 'train.csv'))
train_df['hash'] = np.zeros(44703)
train_df.head()
```

```python
import hashlib

hashes = [];
for record in tqdm(train_df['image_id']):
    with open(IMAGE_FOLDER+record, "rb") as f:
        hash = hashlib.sha256(f.read()).hexdigest()
    hashes.append(hash)
train_df['hash'] = hashes
```

```python
items = len(train_df['image_id'])
uniques = train_df['hash'].nunique()
train_df = train_df.reset_index()
grouped_hash = train_df.groupby('hash')
```

```python
grouped_hash_count = grouped_hash['hash'].count()
dups = grouped_hash_count[grouped_hash_count > 1]
```

```python
# print(dups.head())
duplicate_ids = train_df[train_df['hash'].isin(dups.index)]
print(duplicate_ids.sort_values(by=['hash','hotel_id']))
print(duplicate_ids.head())
```

```python
test_img1_id = duplicate_ids['image_id'].iloc[0]
test_img1_hotel = duplicate_ids['hotel_id'].iloc[0]
print(test_img1_hotel)
test_img2_id = duplicate_ids['image_id'].iloc[1]
test_img2_hotel = duplicate_ids['hotel_id'].iloc[1]
print(test_img2_hotel)
```

```python
test_image = np.array(pil_image.open('../input/hotel-id-to-combat-human-trafficking-2022-fgvc9/train_images/'+str(test_img1_hotel)+'/'+str(test_img1_id))).astype(np.uint8)
plt.figure(figsize=(6,6))
plt.imshow(test_image)
```

```python
test_image = np.array(pil_image.open('../input/hotel-id-to-combat-human-trafficking-2022-fgvc9/train_images/'+str(test_img2_hotel)+'/'+str(test_img2_id))).astype(np.uint8)
plt.figure(figsize=(6,6))
plt.imshow(test_image)
```
