```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

def showRGBImageList(imagePath):
    _, ax = plt.subplots(1, 4, figsize=(22,6))
    
    imageArray = np.array(Image.open(imagePath)).astype(np.uint8)
    
    ax[0].imshow(imageArray)
    ax[0].set_title("Original")
    
    channels = ["Red", "Green", "Blue"]
    for i in range(3):
        ax[i+1].imshow(imageArray[:, :, i])
        ax[i+1].set_title(channels[i])
```

```python
TEST_IMAGE_PATH = "../input/hotel-id-to-combat-human-trafficking-2022-fgvc9/test_images/abc.jpg"
showRGBImageList(TEST_IMAGE_PATH)
```
