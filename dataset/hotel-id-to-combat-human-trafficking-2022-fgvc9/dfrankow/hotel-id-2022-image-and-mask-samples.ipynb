# Introduction

Sample training images and test masks, so I can easily download them locally to play with them.

```python
import os
import glob
```

```python
NUM_TRAIN_IMAGES=50
NUM_TEST_MASKS=50
```

```python
data_folder = "/kaggle/input/hotel-id-to-combat-human-trafficking-2022-fgvc9/"
train_folder = os.path.join(data_folder, 'train_images')
# yep, test masks are in the "train_masks" folder
# see https://www.kaggle.com/competitions/hotel-id-to-combat-human-trafficking-2022-fgvc9/discussion/313547
test_mask_folder = os.path.join(data_folder, 'train_masks')
```

```python
train_image_paths = glob.glob(train_folder + "/*/*")
test_mask_paths = glob.glob(test_mask_folder + "/*")
```

```python
print(len(train_image_paths))
print(len(test_mask_paths))
```

```python
import random

train_image_paths_sample = random.sample(train_image_paths, NUM_TRAIN_IMAGES)
test_mask_paths_sample = random.sample(test_mask_paths, NUM_TEST_MASKS)
```

```python
print(len(train_image_paths_sample))
print(len(test_mask_paths_sample))

# print(train_image_paths_sample)
# print(test_mask_paths_sample)
```

```python
!cd /kaggle/working
!rm -rf train_sample
!rm -rf test_mask_sample
!mkdir train_sample
!mkdir test_mask_sample
```

```python
import pathlib
import shutil
```

```python
root_destdir = '/kaggle/working/train_sample'

for path in train_image_paths_sample:
    the_path = pathlib.PurePath(path)
    the_chain = the_path.parts[-2]
    destdir = root_destdir + '/' + the_chain
    if not os.path.isdir(destdir):
        # print(f"make {destdir}")
        os.mkdir(destdir)
    shutil.copy(path, destdir)
```

```python
destdir = '/kaggle/working/test_mask_sample'

for path in test_mask_paths_sample:
    # print(f"copy {path} to {destdir}")
    shutil.copy(path, destdir)
```
