# Importing the data

```python
df_stock_list = pd.read_csv('../input/jpx-tokyo-stock-exchange-prediction/stock_list.csv')
```

```python
df_stock_list.head()
```

# Various Sectors in the stock market

```python
# Replace the `-` data with unknown
df_stock_list['17SectorName'] = df_stock_list['17SectorName'].replace('-','UNKNOWN')

# Calculating the frequency of occurance of each stock
values = df_stock_list['17SectorName'].value_counts()

# Extracting labels for each sector name
labels = df_stock_list['17SectorName'].value_counts().index.tolist()
```

## Visualization of presence of various sectors

Lets see the top occurances for of various sectors

```python
values.head(30)
```

### Pie Chart for Sectors

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams["figure.figsize"] = (20,15)
data = values
colors = sns.color_palette('pastel')
plt.pie(data,labels=labels, colors = colors)
plt.show()
```

# Various Sub Sectors in the stock market

```python
# Replace the `-` data with unknown
df_stock_list['33SectorName'] = df_stock_list['33SectorName'].replace('-','UNKNOWN')

# Calculating the frequency of occurance of each stock
values_sub = df_stock_list['33SectorName'].value_counts()

# Extracting labels for each sector name
labels_sub = df_stock_list['33SectorName'].value_counts().index.tolist()
```
