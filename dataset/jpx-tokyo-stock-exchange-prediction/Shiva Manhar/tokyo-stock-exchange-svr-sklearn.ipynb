```python
from sklearn import svm, preprocessing
import math
from sklearn import preprocessing
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
stock_prices = pd.read_csv('/kaggle/input/jpx-tokyo-stock-exchange-prediction/train_files/stock_prices.csv')
```

```python
secondary_stock = pd.read_csv('/kaggle/input/jpx-tokyo-stock-exchange-prediction/train_files/secondary_stock_prices.csv')
```

```python
stock_1301 = stock_prices[stock_prices['SecuritiesCode'] == 1376]
```

```python
stock_1301.shape
```

```python
stock_1301
```

```python
stock_1301.set_index('Date')
```

```python
stock_1301 = stock_1301[['Open', 'High', 'Low', 'Close', 'Volume', 'Target']]
```

```python
stock_1301.head(5)
```

```python
stock_1301['HL_PCT']= (stock_1301['High'] - stock_1301['Low'])/stock_1301['Close']*100.0
```

```python
stock_1301['PCT_change'] = (stock_1301['Close'] - stock_1301['Open']) / stock_1301['Open'] * 100.0
```

```python
df = stock_1301[['Close', 'HL_PCT', 'PCT_change', 'Volume', 'Target']]
```

```python
forecast_out = int(math.ceil(0.01 * len(df)))
forecast_out
```

```python
forecast_col = 'Close'
df['label'] = df[forecast_col].shift(-forecast_out)
```

```python
df.head()
```

```python
df.dropna(inplace=True)
X = np.array(df.drop(['label'], 1))
y = np.array(df['label'])
```

```python
X = preprocessing.scale(X)
```

```python
y = np.array(df['label'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)
```

```python
clf = svm.SVR()
```

```python
clf.fit(X_train, y_train)
```

```python
confidence = clf.score(X_test, y_test)
print(confidence)
```
