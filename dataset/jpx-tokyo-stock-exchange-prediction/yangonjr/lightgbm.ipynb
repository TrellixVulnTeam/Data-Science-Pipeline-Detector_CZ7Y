```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import lightgbm as lgb
```

```python
df = pd.read_csv('narma_sim.csv')
```

```python
df_index = df.columns.values
```

```python
# 前処理
X = df[df_index[1:4]]
Y = df[df_index[-1]]
```

```python
# 訓練データとバリデーションデータに分割
X_train, X_validation, y_train, y_validation = train_test_split(X, Y, test_size=0.2, shuffle=False)
```

```python
SEED = 42

model = lgb.LGBMRegressor(
    random_state = SEED,
)

model.fit(
    X_train, 
    y_train,
    eval_set=[(X_validation, y_validation), (X_train, y_train)],
)

# 学習履歴の表示
lgb.plot_metric(model)
```

```python
train_pred = model.predict(X_train)
test_pred = model.predict(X_validation)
```

```python
plt.figure(figsize = (20,10))
plt.plot(np.array(Y), label="raw")
plt.plot(train_pred, label="train")
plt.plot(range(train_pred.shape[0], train_pred.shape[0] + test_pred.shape[0]), test_pred, label="test")
plt.legend()
plt.show()
```
