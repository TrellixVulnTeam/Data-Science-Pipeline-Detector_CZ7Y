```python
# Feature Selection -2nd Model 
X = df.drop(['EffectiveDate','NewMarketSegment'],axis=1)
y=df['NewMarketSegment']
```

```python
# Training & Testing of Model using SK Learn 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2,random_state=0)
```

```python
# Checking Shape of  Model 
X.shape
y.shape
X_train
y_train
```

```python
# Scaling of Data for better performance- Import Standard Scaler 
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
df_std = df['NewMarketSegment'].values.reshape(-1,1)
df_std = std.fit_transform(df_std)
df_std = pd.DataFrame(df_std)
df['NewMarketSegment'] = df_std
```

```python
from matplotlib.cbook import print_cycles
print(df)
```

```python
#import the regressor
from sklearn.tree import DecisionTreeRegressor 
  
# create a regressor object
regressor= DecisionTreeRegressor(random_state = 0) 
  
# fit the regressor with X and Y data
regressor.fit(X, y)
```

```python
#Train the model
regressor.fit(X_train,y_train)
```

```python
#Test/Predict
y_pred = regressor.predict(X_test)
```

```python
y_pred
```

```python
# Reshape the Test Value for better performance 

y_test = y_test.values.reshape(-1,1)
```

```python
# Accuracy -Use score method to get accuracy of model
Accuracy_score = regressor.score(X_test, y_test)*100
print(Accuracy_score)
```

```python
# Comparison Actual Score and Predicted Scor
df_comp = pd.DataFrame({'Actual_NewMarketSegment':y_test.flatten(),'Predicted_NewMarketSegment':y_pred.flatten()})
```

```python
# Confusion Matrix to determine True Posative / True Negative and False Posative / False Negative 

plt.figure(figsize=(9,9))
sns.heatmap(cf, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r');
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'AccuracyScore: {0}'.format(Accuracy_score)
plt.title(all_sample_title, size = 15);
```
