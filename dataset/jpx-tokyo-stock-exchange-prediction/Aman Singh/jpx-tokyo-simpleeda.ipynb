## Basic EDA

```python
stocks = pd.read_csv("/kaggle/input/jpx-tokyo-stock-exchange-prediction/train_files/stock_prices.csv")
```

```python
stocks
```

```python
stocks['Target'].hist(bins = 25, figsize = (20,5))
```

The target values look quite normal without any outliers or long tails. We should not have any problems working with it. Let's also plot distributions of targets of a few random features:

```python
for f in np.random.choice(stocks['SecuritiesCode'].unique(), 10):
    stocks[stocks['SecuritiesCode'] == f]['Target'].hist(bins = 50, alpha = 0.2, figsize = (20,10))
```

On a high-level target for each investment_id also looks ok.
