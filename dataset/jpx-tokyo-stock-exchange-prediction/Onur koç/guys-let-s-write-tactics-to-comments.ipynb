
## 2-) Rolling factor should not be very small for SMA rolling 2,5,10,15,2. 

## Rolling factor should be more than 100, for ex 100,200,500,1000

```python
def calc_change_rate_base2(df,column_name, periods):
    
        for period in periods:
            df.loc[:, f"{column_name}_MOM_{period}"] = np.log(df[column_name]).diff().rolling(window=period, min_periods=1).std()
        return df

df = calc_change_rate_base2(df,"Close",[*range(100,1000,100)])
```
