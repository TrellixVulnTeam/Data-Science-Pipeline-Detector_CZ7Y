This is yet another LGB notebook in the JPX competition

Its main difference is the modular structure:

- An utility script [jpx_feat_util](https://www.kaggle.com/code/notabene/jpx-feat-util) deals with the pre-processing & feature engineering
- This script is used by a notebook [jpx_feat_generation_no_leak](https://www.kaggle.com/code/notabene/jpx-feat-generation-no-leak) which deals with the CV. It pre-computes the CV fold files which can then be used by simply importing the output of this notebook
- Another notebook [jpx_lgb_train_no_leak](https://www.kaggle.com/code/notabene/jpx-lgb-train-no-leak) imports the fold files and trains a separate LGB model on each fold
- The current notebook then imports the models and the utility script to produce a submission file

```python
import numpy as np
import pandas as pd
import lightgbm
import pickle
import jpx_tokyo_market_prediction
from jpx_feat_util import get_stock_features, create_features, drop_cols, fill_nan_inf, inference
import warnings
warnings.filterwarnings('ignore')
```

```python
env = jpx_tokyo_market_prediction.make_env()
iter_test = env.iter_test()

supp = pd.read_csv('../input/jpx-tokyo-stock-exchange-prediction/supplemental_files/stock_prices.csv').tail(2000*68)
supp,models = supp[supp.Open.notnull() & supp.Target.notnull()],[]
for i in range(10):
    models.append(pickle.load(open(f'../input/jpx-lgb-train-no-leak/lgb_model{i}', "rb")))
```

```python
for (prices, options, financials, trades, secondary_prices, sample_prediction) in iter_test:
    env.predict(inference(prices,supp,sample_prediction, models,drop_cols))
```
