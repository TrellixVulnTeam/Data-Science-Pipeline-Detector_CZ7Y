```python
import numpy as np 
import pandas as pd
from jpx_feat_util import get_stock_features, create_features

prices = pd.read_csv('../input/jpx-tokyo-stock-exchange-prediction/train_files/stock_prices.csv')
supp = pd.read_csv('../input/jpx-tokyo-stock-exchange-prediction/supplemental_files/stock_prices.csv')
prices = pd.concat([prices, supp])

consecDates = prices['Date'].unique()
consecDates = sorted(consecDates)
test0, train0 = consecDates[:250], consecDates[320:] #remember to first throw the first 70
test1, train1 = consecDates[250:500], consecDates[:180] + consecDates[570:]
test2, train2 = consecDates[500:750], consecDates[:430] + consecDates[820:]
test3, train3 = consecDates[750:1000], consecDates[:680] + consecDates[1070:]
test4, train4 = consecDates[1000:], consecDates[:930]

prices.shape
```

```python
i = 0
for train, test in [(train0, test0), (train1, test1), (train2, test2), (train3, test3), (train4, test4)]:
    print(f'Creating fold {i}..')
    feats[feats['Date'].isin(frozenset(train))].to_pickle(f"train{i}.pkl")
    feats[feats['Date'].isin(frozenset(test))].to_pickle(f"test{i}.pkl")
    i += 1
```
