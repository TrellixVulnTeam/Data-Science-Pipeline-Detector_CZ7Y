```python
import os
import pandas as pd

# variables
# dataset_dir = '../data/input/'
dataset_dir='../input/jpx-tokyo-stock-exchange-prediction/'
stock_list = None
specification = {}
example_test = {}
supplemental = {}
train = {}

# stock_list
stock_list = pd.read_csv(f'{dataset_dir}/stock_list.csv')

# data_specifications
for dirname, _, filenames in os.walk(os.path.join(dataset_dir, 'data_specifications')):
    for filename in filenames:
        specification[filename.replace('.csv','')] = pd.read_csv(os.path.join(dirname, filename))

# example_test_files
for dirname, _, filenames in os.walk(os.path.join(dataset_dir, 'example_test_files')):
    for filename in filenames:
        example_test[filename.replace('.csv','')] = pd.read_csv(os.path.join(dirname, filename))

# supplemental_files
for dirname, _, filenames in os.walk(os.path.join(dataset_dir, 'supplemental_files')):
    for filename in filenames:
        supplemental[filename.replace('.csv','')] = pd.read_csv(os.path.join(dirname, filename))

# train_files
for dirname, _, filenames in os.walk(os.path.join(dataset_dir, 'train_files')):
    for filename in filenames:
        train[filename.replace('.csv','')] = pd.read_csv(os.path.join(dirname, filename))
```
