```python
import jpx_tokyo_market_prediction

env = jpx_tokyo_market_prediction.make_env()   # initialize the environment
iter_test = env.iter_test()    # an iterator which loops over the test files
```

```python
(prices, options, financials, trades, secondary_prices, sample_prediction) = next(iter_test)
env.predict(sample_prediction)
print(len(prices), len(options), len(financials), len(trades), len(secondary_prices), len(sample_prediction))
df = [prices, options, financials, trades, secondary_prices, sample_prediction]
[d for d in df]
date = prices.at[0,'Date']
prices.to_csv(f'./{date} prices.csv', encoding='utf_8_sig')
options.to_csv(f'./{date} options.csv', encoding='utf_8_sig')
financials.to_csv(f'./{date} financials.csv', encoding='utf_8_sig')
trades.to_csv(f'./{date} trades.csv', encoding='utf_8_sig')
secondary_prices.to_csv(f'./{date} secondary_prices.csv', encoding='utf_8_sig')
sample_prediction.to_csv(f'./{date} sample_prediction.csv', encoding='utf_8_sig')
```
