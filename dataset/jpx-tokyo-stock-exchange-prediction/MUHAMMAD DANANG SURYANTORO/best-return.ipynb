# Find Securities with best return

From dataset we have that 2000 alternative securities. but, we want to choose that best securities for invesment. I think that we must consider from highest return from periode, average return per day or minimum drawdown per days.

```python
temp = {}
calc_gain = pd.DataFrame(columns = ['stokid','meanperday','totalgain'])
for stok in stock_prices.SecuritiesCode.unique():
    temp[stok] = stock_prices[stock_prices['SecuritiesCode']==stok]
    calc_gain = calc_gain.append({'stokid' : stok, 'meanperday': temp[stok].Target.mean(),'totalgain': temp[stok].Target.sum()}, ignore_index = True)
    
```

```python
calc_gain.sort_values(by = ['meanperday','totalgain'], ascending = False)
```
