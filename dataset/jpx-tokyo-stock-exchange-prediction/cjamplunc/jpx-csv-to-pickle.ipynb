```python
import os
import pandas as pd
```

```python
def csv_to_pickle(input_dir,output_dir):
    for dirname, _, filenames in os.walk(input_dir):
        filenames = [v for v in filenames if ".csv" in v]
        if not filenames:
            continue

        out_dirname = dirname.replace(input_dir,output_dir)
        if not os.path.isdir(out_dirname):
            os.makedirs(out_dirname)

        for filename in filenames:
            df = pd.read_csv(os.path.join(dirname, filename))
            output_path = os.path.join(out_dirname, filename).replace(".csv",".p")
            df.to_pickle(output_path)
            print(f"Created {output_path}")
```

```python
%%time
input_dir = "/kaggle/input/jpx-tokyo-stock-exchange-prediction/"
output_dir = "/kaggle/working/"

csv_to_pickle(input_dir,output_dir)
```
