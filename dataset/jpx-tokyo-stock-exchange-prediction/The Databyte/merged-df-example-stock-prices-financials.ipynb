```python
import pandas as pd 
import numpy as np

path = "../input/jpx-tokyo-stock-exchange-prediction/"
dfsp = pd.read_csv(f"{path}train_files/stock_prices.csv")
dff = pd.read_csv(f"{path}train_files/financials.csv")
```

```python
#merge the two dataframes together on similar columns (Date, Securities Code)
df_merged = pd.merge(dfsp, dff, on=['Date', 'SecuritiesCode'], how='outer')
```
