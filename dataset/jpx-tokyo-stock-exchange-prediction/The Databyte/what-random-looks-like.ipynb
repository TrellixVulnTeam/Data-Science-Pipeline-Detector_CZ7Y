```python
import pandas as pd 
import numpy as np
```

```python
#pulling in sample submission file for stock tickers
path = "../input/jpx-tokyo-stock-exchange-prediction/"
df = pd.read_csv(f"{path}example_test_files/sample_submission.csv")
```

```python
#creating a column of random data between 0 & 1
df['Rank'] = np.random.uniform(0, 1, size=len(df))
print(df)
```

```python
#filtering to one date to remove duplicate tickers
dfsmall = df.loc[df['Date'] == '2021-12-06']
#print(dfsmall)

#sorting the values
dfsmall = dfsmall.sort_values("Rank", ascending=False).drop_duplicates(subset=['SecuritiesCode'])
dfsmall.loc[:, "Rank"] = np.arange(len(dfsmall))
dfsmall.tail()

#creating a dictionary to map values
rank = dfsmall.set_index('SecuritiesCode')['Rank'].to_dict()
#print(rank)
```

```python
import jpx_tokyo_market_prediction

env = jpx_tokyo_market_prediction.make_env()
iter_test = env.iter_test()

for (prices, _, _, _, _, submission) in iter_test:
    submission["Rank"] = submission["SecuritiesCode"].map(rank)
    env.predict(submission)
```
