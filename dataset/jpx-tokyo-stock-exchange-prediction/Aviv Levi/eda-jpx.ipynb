## Let's first look at the main file -  stock_prices.csv

```python
prices.head()
```

### Let's see what is tha nan's ratio


```python
prices_len = prices.shape[0]
nans = prices.isna().sum() / prices_len
nans.round(3).sort_values(ascending = False)
```

1. ExpectedDividend is almost 100% nans
2. ### Open High Low Close have the same nans pretty much. prehaps nan at one means nan at all of those?

```python
nans_df = prices.loc[prices["Open"].isna()]
OHLC_nans = nans_df[["Open", "High", "Low", "Close"]]
OHLC_nans
```

```python
print("number of not nans, given Open is nan")
OHLC_nans.isna().sum() - OHLC_nans.shape[0]
```

### Ok, so those look like nan prices. are the targets nan as well?

```python
print("% of nan targets if price has nans in Open, High, Low, Close")
print(nans_df["Target"].isna().sum() / nans_df.shape[0])
```

### So there is about 3% missing targets given one of  Open, High, Low, Close is nan
