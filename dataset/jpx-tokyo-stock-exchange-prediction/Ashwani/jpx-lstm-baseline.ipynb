## Lets visualize stock prices
Based on the closing price

```python
SecuritiesCode = 2002 # change the Securities Code for the stock you want to visualize
```

```python
stock_prices[stock_prices['SecuritiesCode'] == SecuritiesCode].head(3)
```

```python
def visualize_stocks(code):
    try:
        stock_code = stock_prices[stock_prices['SecuritiesCode'] == code][['Date', 'Close']]
        fig = px.line(stock_code, x='Date', y='Close', labels={
                             "Close": "Closing Price",
                             "Date": "Date (years)"
                         },
                        title=stock_list[stock_list['SecuritiesCode']==code]['Name'].values[0])

        fig.show()
    except:
        print("This Securities code is not Valid (please check and enter again. . .)")
```

```python
visualize_stocks(SecuritiesCode)
```

```python
def visualize_profit(code):
    try:
        profits = financials[financials['SecuritiesCode']==code].reset_index(drop=True)[['Date','OperatingProfit', 'OrdinaryProfit', 'Profit']]
        profits.dropna(inplace=True)
        profits.drop_duplicates(inplace=True, subset=['OperatingProfit', 'OrdinaryProfit', 'Profit'])
        profits[['OperatingProfit', 'OrdinaryProfit', 'Profit']] = profits[['OperatingProfit', 'OrdinaryProfit', 'Profit']].astype(int)
        profits['Year'] = pd. DatetimeIndex(profits['Date']).year
        profits_yearly = profits.groupby('Year')['Profit'].sum().reset_index()
        fig = px.bar(profits_yearly, x="Year", y="Profit",labels={
                                     "Profit": "Profit (yearly)",
                                     "Year": "Date (years)"
                                 },
                                title="Yearly Profits for: "+stock_list[stock_list['SecuritiesCode']==code]['Name'].values[0])
        fig.show()
    except:
        print("This Securities code is not Valid (please check and enter again. . .)")
```

```python
visualize_profit(SecuritiesCode)
```
