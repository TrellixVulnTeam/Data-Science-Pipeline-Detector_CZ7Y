# JPX Tokyo Stock Exchange Data View

```python
import os
import pandas as pd
```

# a) data_specifications

```python
paths=[]
names=[]
for dirname, _, filenames in os.walk('../input/jpx-tokyo-stock-exchange-prediction/data_specifications'):
    for filename in filenames:
        paths+=[os.path.join(dirname, filename)]
        names+=[filename[0:-4]]
names2a=[]
for i,item in enumerate(names):
    names2a+=['a'+str(i)+'_'+item]

print(names2a)
print()
names3a=names2a.copy()

for i in range(len(names2a)):
    names3a[i]=pd.read_csv(paths[i])
    print(names2a[i])
    display(names3a[i][0:2].T)
    print()
    print()
    
```
