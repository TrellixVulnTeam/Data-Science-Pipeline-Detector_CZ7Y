## stock_prices.csv

*/stock_prices.csv The core file of interest. Includes the daily closing price for each stock and the target column.

* `RowId`: Unique ID of price records, the combination of `Date` and `SecuritiesCode`.
* `Date`: Trade date.
* `SecuritiesCode`: Local securities code.
* `Open`: First traded price on a day.
* `High`: Highest traded price on a day.
* `Low`: Lowest traded price on a day.
* `Close`: Last traded price on a day.
* `Volume`: Number of traded stocks on a day.
* `AdjustmentFactor`: Used to calculate theoretical price/volume when split/reverse-split happens (NOT including dividend/allotment of shares).
* `ExpectedDividend`: Expected dividend value for ex-right date. This value is recorded 2 business days before ex-dividend date.
* `SupervisionFlag`: Flag of securities under supervision and securities to be delisted, for more information, please see [here](https://www.jpx.co.jp/english/listing/market-alerts/supervision/00-archives/index.html).
* `Target`: Change ratio of adjusted closing price between t+2 and t+1 where t+0 is trade date.

**Targetは下の式に基づいて導出されます。実際にいくつかのデータで導出してみましょう**

**Target is derived based on the formula below. Let's actually derive from some data**

$$r_{(k, t)} = \frac{C_{(k, t+2)} - C_{(k, t+1)}}{C_{(k, t+1)}}$$

```python
df = df_train_stock[df_train_stock['SecuritiesCode'] == 1301]
df["Close_shift1"] = df["Close"].shift(-1)
df["Close_shift2"] = df["Close"].shift(-2)

df["rate"] = (df["Close_shift2"] - df["Close_shift1"]) / df["Close_shift1"]
df
```

**Close_shift1は次の日のClose、Close_shift2は次の次の日のCloseです。上の式によって導出されたrateとTargetが一致していることがわかります。**

**Close_shift1 is the next day's Close, and Close_shift2 is the next day's Close. You can see that the rate and Target derived by the above formula match.**

```python
def plot_candle_with_target(stock_code, prime=True):
    """Plot OHLCV plot with target series.
    
    Parameters:
        stock_code: int, code of the stock
        prime: bool, whether the stock to plot is prime or not
    """
    df_ = df_train_stock.copy()
    df_ = df_[df_['SecuritiesCode'] == stock_code]
    dates = df_['Date'].values
    ohlc = {
        'open': df_['Open'].values, 
        'high': df_['High'].values, 
        'low': df_['Low'].values, 
        'close': df_['Close'].values
    }
    vol = df_['Volume'].values
    target = df_['Target'].values
    
    fig = make_subplots(rows=3, cols=1, shared_xaxes=True, x_title='Date')
    fig.add_trace(go.Candlestick(x=dates, name='OHLC', **ohlc),
                  row=1, col=1)
    fig.add_trace(go.Bar(x=dates, y=vol, name='Volume'),
                  row=2, col=1)
    fig.add_trace(go.Scatter(x=dates, y=target, name='Target'),
                  row=3, col=1)
    fig.update_layout(
        title=f"OHLCV Chart with Target Series (Stock {stock_code})",
    )
    fig.update(layout_xaxis_rangeslider_visible=False)
    fig.show()
```

**このコードを使うと、ある日から20日分のデータを取り出すことができます。これを使って、範囲を狭めたグラフを表示する関数を書いてみます。**

```python
def plot_candle_with_target_ndays(stock_code,target_day,ndays, prime=True):
    """Plot OHLCV plot with target series.
    
    Parameters:
        stock_code: int, code of the stock
        prime: bool, whether the stock to plot is prime or not
    """
    df_ = df_train_stock.copy()
    df_ = df_[df_['SecuritiesCode'] == stock_code].reset_index()
    s=df_.index[df_['Date']==target_day].tolist()
    if len(s)==0 or len(df_)-s[0]<ndays:
        return 
    df_ = df_.loc[s[0]:s[0]+ndays]
    dates = df_['Date'].values
    ohlc = {
        'open': df_['Open'].values, 
        'high': df_['High'].values, 
        'low': df_['Low'].values, 
        'close': df_['Close'].values
    }
    vol = df_['Volume'].values
    target = df_['Target'].values
    
    fig = make_subplots(rows=3, cols=1, shared_xaxes=True, x_title='Date')
    fig.add_trace(go.Candlestick(x=dates, name='OHLC', **ohlc),
                  row=1, col=1)
    fig.add_trace(go.Bar(x=dates, y=vol, name='Volume'),
                  row=2, col=1)
    fig.add_trace(go.Scatter(x=dates, y=target, name='Target'),
                  row=3, col=1)
    fig.update_layout(
        title=f"OHLCV Chart with Target Series (Stock {stock_code})",
    )
    fig.update(layout_xaxis_rangeslider_visible=False)
    fig.show()
```

```python
import random
from plotly.subplots import make_subplots
import plotly.graph_objects as go
code_list = random.sample(df_train_stock['SecuritiesCode'].unique().tolist(), len(df_train_stock['SecuritiesCode'].unique()))
for stock_code in code_list[:3]:
    plot_candle_with_target(stock_code)
    
ndays = 50
target_day='2020-11-26'
for stock_code in code_list[:3]:
    plot_candle_with_target_ndays(stock_code,target_day,ndays)
```
