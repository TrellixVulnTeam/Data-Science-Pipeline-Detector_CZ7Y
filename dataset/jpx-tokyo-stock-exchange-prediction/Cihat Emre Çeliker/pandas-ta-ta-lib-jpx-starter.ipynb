```python
import pandas as pd
import pandas_ta as ta

df = pd.read_csv("../input/jpx-tokyo-stock-exchange-prediction/supplemental_files/secondary_stock_prices.csv")
df = df[df.SecuritiesCode==1321]
df.sample()
```

#### Adding all candlestick patterns.

```python
df = pd.concat([df, df.ta.cdl_pattern()], axis=1)
df[[f for f in df if "CDL" in f]].head()
```

#### There are many more, potentially useful and more straightforward to use indicators you can find in `pandas-ta`  compared to `ta-lib`.

```python
df = pd.concat([df, df.ta.ichimoku(lookahead=False)[0]], axis=1)
df[["Close", "ITS_9", "IKS_26"]][-100:].plot(figsize=(20,5));
```
