```python
paths_dictionary = {}
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        paths_dictionary[filename.split('.')[0]] = os.path.join(dirname, filename)
        
```

```python
from IPython.display import display, Markdown, Latex
```

```python
table = '''| Name | Path |
|:---|:---|
'''
for name, path in paths_dictionary.items():
#     print(name, '\t\t\t: ', path)
        table = table + ' | ' + name + ' | ' + path + ' | \n'
    
display(Markdown(table))
```
