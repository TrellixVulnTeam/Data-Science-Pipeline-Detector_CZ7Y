```python
## Importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
import os
import cv2
import pickle
from keras.utils import to_categorical
```

```python
## Defining the folder to import the images from main kaggle directory
main_path = "/kaggle/input/severstal-steel-defect-detection"
sub_folder = ["train_images"]
```

```python
## Defining the file directory to read the respective folder.

IMAGE_SIZE=100

data=[]

for i in sub_folder:
    folder = os.path.join(main_path,i)
    label = sub_folder.index(i)
    for j in os.listdir(folder):
        path = os.path.join(folder,j)
        img_arr = cv2.imread(path)
        img_arr = cv2.resize(img_arr , (IMAGE_SIZE,IMAGE_SIZE))
        data.append([img_arr,label])
len(data)
```

```python
### Shufflling the data 
random.shuffle(data)
```

```python
data[0]
```

```python
### Defining x and y
x=[]
y=[]
for features,labels in data:
    x.append(features)
    y.append(labels)
        
```

```python
### Converting the x and y into array
x=np.array(x)
y=np.array(y)
```

```python
x
```

```python
### Normalizing the data
x=x/255
```

```python
### Doing labelling on depedent variables
y_one_hot=to_categorical(y)
```

```python
### Doing train test spilt
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
### Importing the important layers related to model building
from keras.models import Sequential
from keras.layers import Dense,Dropout,MaxPooling2D,Flatten,Conv2D
import tensorflow as tf
from tensorflow.keras import models,layers
```

```python
## Defining the model
model=Sequential()
## Conv layer-1
model.add(Conv2D(64,(3,3),activation="relu"))
### Pooling layer-1
model.add(MaxPooling2D((2,2)))
## Conv layer-2
model.add(Conv2D(64,(3,3),activation="relu"))
### pooling layer-2
model.add(MaxPooling2D((2,2)))
### Flatten layer
model.add(Flatten())
### Hidden layer
model.add(Dense(128,activation="relu",input_shape=(100,100,3)))
### Output layer
model.add(Dense(4,activation="softmax"))
```

```python
### compling the model
model.compile(loss="sparse_categorical_crossentropy",
             metrics=["accuracy"],
             optimizer="adam")
```

```python
### Fitting the model-1(simply without labelling)
model.fit(x,y,epochs=5,validation_split=0.2)
```

```python
### Fitting the model with y_categorical
model.fit(x,y_one_hot,epochs=5,validation_split=0.2)
```

```python
### Fitting the model with y_categorical
model.fit(x_train,y_train_one_2,epochs=5,validation_split=0.2)
```
