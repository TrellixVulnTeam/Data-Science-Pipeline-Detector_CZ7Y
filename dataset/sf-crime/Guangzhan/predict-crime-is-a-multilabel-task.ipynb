# Predict crime is a multilabel task

Ok, probably this is not the most fancy title, but this script does really want to show only that the task of this challenge
is most probably a multiclass/multilabel classification task.

Take for example the very first two lines of the training set:

```python
print ("hllo work")
```

How can you distinguish the two, if you remove the labels? [*You can't*].

If you take out Category and Description from those two records, the events are exactly the same! This probably occurs when the same event is linked to more categories or prosecutions, like two different kind of violations (like in the example) or more than one person arrested.

So let's see how many of these cases exist:

```python
%matplotlib inline
counts = train.groupby(['Dates', 'X', 'Y']).size()

counts.value_counts().plot('bar', logy=True)
counts.value_counts()
```

From the plot above we see that having two events exactly at the same time and place is not so rare...Multiple arrests? Probably.
Multiple crimes? Maybe.

Anyway, a classificator should be aware of this possibility to do its tasks at its best.

Following, some examples of the above multiple events, for the curious.

```python
other = pd.DataFrame(counts[counts>=13])
other = other.reset_index()
manyarrests = train.merge(other, how='right')
```
