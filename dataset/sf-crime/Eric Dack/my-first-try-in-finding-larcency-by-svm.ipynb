```python
import numpy as np 
import pandas as pd 
import zipfile
from sklearn import svm

drive = '../input/'

# Read data
z = zipfile.ZipFile(drive+'train.csv.zip')
data = pd.read_csv(z.open('train.csv'))

# Change string to number
crimeSet = data.Category.unique()
i=1
for crime in crimeSet:
    data.Category[data.Category == crime] = i
    i = i+1

daySet = data.DayOfWeek.unique()
i=1
dayInWeek =['Sunday','Monday','Tuesday','Wednesday','Thursday','Friday','Saturday'] 
for day in daySet:
    data.DayOfWeek[data.DayOfWeek == day] = dayInWeek.index(day)
    
districtSet = data.PdDistrict.unique()
i=1
for district in districtSet:
    data.PdDistrict[data.PdDistrict == district] = i
    i = i+1

data['X'] = list([round(val*10.0) for val in data.X.values])
data['Y'] = np.array([round(val*10.0) for val in data.Y.values])

# Generate the train set and the validation set
maxIdx = data.shape[0]-1
sampler = np.random.permutation(maxIdx)
trainSet = data.take(sampler[0:int(0.75*maxIdx)])
validationSet = data.take(sampler[int(0.75*maxIdx):])

# 
nCrime = data.Category.unique()
clf = svm.SVC(gamma=0.001, C=len(nCrime))
clf.fit(np.array(trainSet[['DayOfWeek','PdDistrict','X','Y']].values.tolist()),np.array(trainSet['Category'].values.tolist()))
```
