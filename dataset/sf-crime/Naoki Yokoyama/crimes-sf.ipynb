## Features Engineer

```python
#data train
train['mes'] = train['Dates'].apply(lambda x: x[5:7])#mes
train['ano'] = train['Dates'].apply(lambda x: x[0:4])#ano
train['dia'] = train['Dates'].apply(lambda x: x[8:10])#dia
train['hora'] = train['Dates'].apply(lambda x: x[10:13])#hora
train['min'] = train['Dates'].apply(lambda x: x[14:16])#min
```

```python
#data test
test['mes'] = test['Dates'].apply(lambda x: x[5:7])#mes
test['ano'] = test['Dates'].apply(lambda x: x[0:4])#ano
test['dia'] = test['Dates'].apply(lambda x: x[8:10])#dia
test['hora'] = test['Dates'].apply(lambda x: x[10:13])#hora
test['min'] = test['Dates'].apply(lambda x: x[14:16])#min
```

```python
X = train.drop(['Resolution', 'Address', 'Category', 'Dates', 'Descript'], axis=1) 
y = train.Category

test = test.drop(['Address', 'Dates', 'Id'], axis=1)
```

```python
X.shape, test.shape
```

```python
#pre-processamento
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()


X['DayOfWeek'] = le.fit_transform(X['DayOfWeek'])
X['PdDistrict'] = le.fit_transform(X['PdDistrict'])


test['DayOfWeek'] = le.fit_transform(test['DayOfWeek'])
test['PdDistrict'] = le.fit_transform(test['PdDistrict'])

y = le.fit_transform(y)
```

```python
#smote
smote = SMOTE()
X,y = smote.fit_sample(X,y)

```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score
from sklearn.linear_model import LogisticRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
tree = DecisionTreeClassifier(random_state=42)
tree.fit(X_train, y_train)
```

```python
#sem Smote
tree.score(X_train, y_train) , tree.score(X_test, y_test)
```

```python
#com Smote
tree.score(X_train, y_train) , tree.score(X_test, y_test)
```

```python
#predict
y_pred = tree.predict(test)
```

```python
y_pred_dummy = pd.get_dummies(y_pred)
```
