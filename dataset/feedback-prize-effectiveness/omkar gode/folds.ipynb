```python
import pandas as pd
from sklearn.model_selection import StratifiedGroupKFold

cv = StratifiedGroupKFold(5,shuffle=True,random_state=43)

train = pd.read_csv('../input/feedback-prize-effectiveness/train.csv')

train['kfold'] = -1

for k,(_,val_idx) in enumerate( cv.split( train, y=train.discourse_type, groups=train.essay_id ) ):
    
    train.loc[ val_idx, 'kfold' ] = k

train.to_csv('train_folds.csv',index=None)

!cp -r ../input/feedback-prize-effectiveness/* ./
```
