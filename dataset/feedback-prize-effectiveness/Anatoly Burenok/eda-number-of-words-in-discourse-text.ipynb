```python
import pandas as pd

import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = [12, 6]
plt.rcParams["font.size"] = 12
```

```python
df_origin = pd.read_csv("../input/feedback-prize-effectiveness/train.csv")

df_origin.head()
```

```python
use_cols = ['discourse_text', 'discourse_type', 'discourse_effectiveness']

df = df_origin[use_cols].copy()
df['nwords'] = df['discourse_text'].apply(lambda s: len([w for w in s.split()]))

df.head()
```

```python
df.sort_values(by='nwords', ascending=False)
```

```python
use_col = "discourse_effectiveness"

for col in df[use_col].unique():
    col_data = df.loc[df[use_col] == col, 'nwords']
    col_max = col_data.max()

    col_data.plot.hist(bins=col_max)
    plt.xlabel('NWords')
    plt.title(label=f'{col}', color='blue')
    plt.show()
    
    display(col_data.describe().to_frame(f'{col} nwords').T.astype(int))
```

```python
use_col = "discourse_type"

for col in df[use_col].unique():
    col_data = df.loc[df[use_col] == col, 'nwords']
    col_max = col_data.max()

    col_data.plot.hist(bins=col_max)
    plt.xlabel('NWords')
    plt.title(label=f'{col}', color='blue')
    plt.show()
    
    display(col_data.describe().to_frame(f'{col} nwords').T.astype(int))
```

```python
df.boxplot(column='nwords', by='discourse_type');
```

```python
trim_after = 250

trim_mask = df['nwords'] < trim_after
df[trim_mask].boxplot(column='nwords', by='discourse_type');
```

```python
use_col = "discourse_type"
by_col = "discourse_effectiveness"

for col in df[use_col].unique():
    mask = df[use_col] == col
    col_data = df.loc[mask, ['nwords', by_col]]
    
    col_data.boxplot(column='nwords', by=by_col)
    
    plt.ylabel('NWords')
    plt.xlabel(None)
    plt.title(label=f'{col}', fontsize=18, color="blue")
    plt.show()
    
    display(
        col_data.groupby(by_col)['nwords'].describe().astype(int) \
            .rename_axis(f'{col} nwords', axis='columns')
    )
```

```python
trim_after = 250

for col in df[use_col].unique():
    mask = (df[use_col] == col) & (df['nwords'] < trim_after)
    col_data = df.loc[mask, ['nwords', by_col]]

    col_data.boxplot(column='nwords', by=by_col)
    
    plt.ylabel('NWords')
    plt.xlabel(None)
    plt.title(label=f'{col} < {trim_after}', fontsize=18, color="blue")
    plt.show()
    
    display(
        col_data.groupby(by_col)['nwords'].describe().astype(int) \
            .rename_axis(f'{col} nwords', axis='columns'))
```
