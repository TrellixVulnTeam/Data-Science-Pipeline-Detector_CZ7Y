We are tasked to predict the effectiveness of these discourse elements in the essays. There are 7 types of discourse that are being annotated here. The data description for the competition provides more details:

> **Lead** - an introduction that begins with a statistic, a quotation, a description, or some other device to grab the reader’s attention and point toward the thesis.

> **Position** - an opinion or conclusion on the main question

> **Claim** - a claim that supports the position

> **Counterclaim** - a claim that refutes another claim or gives an opposing reason to the position

> **Rebuttal** - a claim that refutes a counterclaim

> **Evidence** - ideas or examples that support claims, counterclaims, or rebuttals.

> **Concluding Statement** - a concluding statement that restates the claims



### **Observations:** 

From the above bar plot we can observe that **Evidence**, **claim** are having highest in number and **counterclaim**, **Rebuttal** as lowest. As we know Lead, claim, evidence and concluding statement are needed and necessary for every essay and based on the topic we can having many claim and evidence as seen above.


<div class='sub-heading'>Univarient analysis on discourse_effectiveness</div>

```python
dis_eff = dict(train_df['discourse_effectiveness'].value_counts())

names = list(dis_eff.keys())
values = list(dis_eff.values())

plt.figure(figsize=(15,7))
plt.bar(range(len(dis_eff)), values,tick_label=names)
plt.xlabel('discourse class')
plt.ylabel('Count of discourse class')
plt.show();
```

<div class='sub-heading'> Discourse Type vs Discourse Effectiveness</div>

```python
import random
import seaborn as sns

cat_features=['discourse_type']
plt.figure(figsize = (18,18))
def rel_tar(df,feat_list,target):
    
        for i in enumerate(feat_list):
             
                plt.subplot(2,2,i[0]+1)
                sns.countplot(x=i[1], data=df, hue=target,palette='flare')
                plt.title (i[1]+f' vs {target}',fontdict={'fontsize':20})
                plt.xlabel(" ")
                plt.ylabel(" ")
                plt.xticks(rotation = 45)
                plt.tight_layout()
                
                
rel_tar(train_df,cat_features,'discourse_effectiveness')         
```

### **Observation:**

For all seven discourse type we have almost same distribution expect for evidence we have more ineffective class as well.

<div class='sub-heading'> Discourse Text examples:</div>

```python
from IPython.core.display import display, HTML

def show_examples_for_discourse_type(discourse_type, topic):
    filt = train_df.query(f'discourse_type == "{discourse_type}"').sample(frac=1, random_state=420)
    display(HTML(
        f"""
        <h4 style="background:#66ccff ;color: white; font-size: 20px; width:20%; padding: 12px 12px;" >{discourse_type }</h4>
        <table>
            <tr>
              <th style="color:black; font-size: 15px", bgcolor='#4d79ff' width=33%>Ineffective</th>
              <th style="color:black; font-size: 15px", bgcolor='#4d79ff' width=33%>Adequate</th>
              <th style="color:black; font-size: 15px", bgcolor='#4d79ff' width=33%>Effective</th>
            </tr>
            <tr>
              <td>{filt.query("discourse_effectiveness == 'Ineffective'").iloc[0].discourse_text}</td>
              <td>{filt.query("discourse_effectiveness == 'Adequate'").iloc[0].discourse_text}</td>
              <td>{filt.query("discourse_effectiveness == 'Effective'").iloc[0].discourse_text}</td>
            </tr>
        </table>
        """
    ))
```

```python
for dt in set(train_df.discourse_type.values):
    show_examples_for_discourse_type(dt, 10)
```
