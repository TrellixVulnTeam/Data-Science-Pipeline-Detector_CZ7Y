```python
import glob
import pandas as pd

csvs = glob.glob("*.csv")
idx = []
preds = []
for csv_idx, csv in enumerate(csvs):
    df = pd.read_csv(csv)
    # sort df by discourse_id
    df = df.sort_values(by=["discourse_id"])
    print(df.head())
    temp_preds = df.drop(["discourse_id"], axis=1).values
    if csv_idx == 0:
        idx = list(df["discourse_id"])
        preds = temp_preds
    else:
        preds += temp_preds

preds = preds / 5
sample_submission = pd.read_csv("../input/feedback-prize-effectiveness/sample_submission.csv")
sample_submission.loc[:, "discourse_id"] = idx
sample_submission.loc[:, "Ineffective"] = preds[:, 0]
sample_submission.loc[:, "Adequate"] = preds[:, 1]
sample_submission.loc[:, "Effective"] = preds[:, 2]
sample_submission.to_csv("submission.csv", index=False)
```
