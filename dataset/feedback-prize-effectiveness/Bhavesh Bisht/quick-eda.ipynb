```python
def add_text(id_):
    with open(os.path.join(train_dir, id_ + '.txt')) as f:
        lines = f.readlines()
    return lines
temp['paragraph'] = temp['essay_id'].apply(lambda x: add_text(x))
temp['paralen'] = temp['paragraph'].apply(lambda x: len(x))
temp.columns = ['essay_id', 'discourse_type_count', 'paragraph', 'paralen']
```

```python
print("There are {} paragprahs where complete annotations are not present".format(temp.query("discourse_type_count < paralen").shape[0]))
```

```python
train = pd.merge(train, temp, on = ['essay_id'], how = 'left')
```
