```python
import numpy as np
import pandas as pd
```

Load the sample submission file

```python
sub = pd.read_csv('../input/tabular-playground-series-nov-2021/sample_submission.csv')
sub.sort_index(inplace = True) #To make sure all the arrays are aligned
```

Load the predictions of the individual models and store them into a 2D array

```python
files = [
            '../input/lightautoml-november-21/submission.csv',
            '../input/make-it-simple/submission.csv',
            '../input/tps-11-21-eda-xgboost-optuna/submission.csv',
            '../input/tps-nov-2021-simple-ann/submission.csv',
            '../input/tps-11-nn-baseline-keras/nn_baseline.csv'
        ]

arr = np.zeros((len(sub), len(files)))

for i, f in enumerate(files):
    f=pd.read_csv(f)
    f.sort_index(inplace = True) #To make sure all the arrays are aligned
    arr[:, i] = f['target'].values
```

Use mean of the predictions of the various models as the final prediction

```python
#sub['target'] = np.mean(arr, axis = 1)

#sub['target'] = np.median(arr, axis = 1)

#Power averaging
#https://www.kaggle.com/c/tabular-playground-series-nov-2021/discussion/284507
p = 2.
sub['target'] = np.mean(arr**p, axis = 1)**(1/p)
```

Save the final submission

```python
sub.to_csv("submission.csv", index=False)
```
