## 100 tests 0.75042 vs 0.75009

```python
pure = pd.read_csv('../input/tps11-postprocess/pure_test.csv')
sub_75042= pd.read_csv('../input/tps11-postprocess/75042.csv')
sub_75009= pd.read_csv('../input/tps11-postprocess/75009.csv')
score_75042, score_75009=0, 0
for i in range(100):
    pure_loop = pure.copy()
    pure_loop = pure_loop.set_index('id')
    flip_idx = np.arange(540000)
    flip_idx=np.random.choice(flip_idx,540000//4, replace=False)
    pure_loop.iloc[flip_idx,0]=(pure_loop.iloc[flip_idx,0]+1)%2
    pure_loop.iloc[flip_idx,0]
    public = pure_loop.iloc[:int(540000*0.19),0]
    private = pure_loop.iloc[int(540000*0.19):,0]
    public_sub_75009 = sub_75009.iloc[:int(540000*0.19),1]
    private_sub_75009 = sub_75009.iloc[int(540000*0.19):,1]
    public_sub_75042 = sub_75042.iloc[:int(540000*0.19),1]
    private_sub_75042 = sub_75042.iloc[int(540000*0.19):,1]
    if roc_auc_score(private,private_sub_75042) > roc_auc_score(private,private_sub_75009):
        score_75042+=1
    else:
        score_75009+=1
print('sub 75042 score:  ', score_75042, '  sub 75009 score:  ', score_75009)
```
