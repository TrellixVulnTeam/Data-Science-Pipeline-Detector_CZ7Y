```python
import datatable as dt
from autogluon.tabular import TabularPredictor
```

```python
train = dt.fread('/kaggle/input/tabular-playground-series-nov-2021/train.csv')
test = dt.fread('/kaggle/input/tabular-playground-series-nov-2021/test.csv')
```

```python
## run model


train['target'] = dt.Frame(train['target'].to_numpy().ravel())

model_autogluon = TabularPredictor(label='target')
model_autogluon.fit(train_data=train.to_pandas().drop("id",axis =1), excluded_model_types=['KNN'], time_limit=2*60*60)

del train['target']
```

```python
model_autogluon.leaderboard()
```

```python
preds_autogluon = model_autogluon.predict_proba(test.to_pandas().drop("id",axis =1))[True]
preds_autogluon
```

```python
test_ids = test['id']
submission_autogluon = dt.Frame(id=test_ids, target=dt.Frame(preds_autogluon))
```

```python
submission_autogluon.to_csv("submission_autogluon.csv")
```
