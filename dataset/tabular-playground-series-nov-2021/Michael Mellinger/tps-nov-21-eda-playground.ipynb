# **<span style="color:#006994;">Load Data</span>**

```python
%time
train_df = pd.read_csv(data_dir / "train.csv", 
#                        nrows=10000,
                       index_col=0,
                      )
test_df = pd.read_csv(data_dir / "test.csv",
#                      nrows=1000,
                     index_col=0,
                     )
sample_submission = pd.read_csv(data_dir / "sample_submission.csv")

print(f"train data: Rows={train_df.shape[0]}, Columns={train_df.shape[1]}")
print(f"test data : Rows={test_df.shape[0]}, Columns={test_df.shape[1]}")
```

<a id="identify-feature-types"></a>

# **<span style="color:#006994;">Identify Feature Types</span>**

```python
features = []
categorical_features = []
numerical_features = []

for feature in test_df.columns:
    features.append(feature)
    if test_df.dtypes[feature] == object or test_df.dtypes[feature] == 'int8':
        categorical_features.append(feature)
    else:
        numerical_features.append(feature)

plt.bar([1,2],[len(categorical_features),len(numerical_features)])
plt.xticks([1,2],('Categorical','Numerical'))
plt.show()

print("Categorical features:", len(categorical_features))
print("Numerical features:", len(numerical_features))
```

## Calculate Target Ratio

```python
percent_ones = train_df.target.sum()/train_df.target.count()
print(f"Class: 0 = {1 - percent_ones:.4f} ; 1 = {percent_ones:.4f}")
```
