```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")

SEED = 42
np.random.seed(SEED)
```

```python
train = pd.read_csv('../input/tabular-playground-series-nov-2021/train.csv')
test = pd.read_csv('../input/tabular-playground-series-nov-2021/test.csv')
submission = pd.read_csv('../input/tabular-playground-series-nov-2021/sample_submission.csv')
print('Quick view of training data: ')
train.head()
```

```python
label = 'target'
print("Summary of class variable: \n", train[label].describe())
```

```python
from autogluon.tabular import TabularPredictor, TabularDataset
```

```python
predictor = TabularPredictor(
    label=label, 
    eval_metric = 'roc_auc', 
    verbosity=3
).fit(
    train, 
    presets='best_quality', 
    time_limit=3600,
    num_bag_folds = 5,
    num_bag_sets = 1,
    num_stack_levels = 1
)
```

```python
results = predictor.fit_summary(show_plot=True)
```

```python
y_predproba = predictor.predict_proba(test)
y_predproba.head(5)
```

```python
predictor.positive_class
```

```python
y_predproba = predictor.predict_proba(test, as_multiclass=False)
```

```python

submission[label] = y_predproba
submission.head()
submission.to_csv('my_submission.csv', index=False)
```
