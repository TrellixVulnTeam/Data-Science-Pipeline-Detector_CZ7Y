<h1 style="font-size:50px;color:#6166B3"><strong>Introduction to </strong><strong style="color:black">Neural Networks</strong></h1>

<p style="font-size:120%">Birds have insipered us to fly, burdock plants inspired Velcro, and nature has inspired countless more inventions. It seems only logical, then, to look at the brain's architecture for inspiration on how to build an intelligent machine. This is the logic that sparked <strong>Artificial Neural Networks (ANN):</strong> <mark>an ANN is a Machine Learning model inspired by the networks of biological neurons found in our brains.</mark> However, although planes were inspired by birds, they don't have to flap their wings. Similarly, ANNs have gradually become quite different from their biological cousins. Some researchers argue that we should drop the biological analogy altogether (e.g., by saying "units" rather than "neurons"), lest we restrict our creativity to biologically plausible systems.</p>

<p style="font-size:120%">ANNs are at the very core of Deep Learning. <mark>They are versatile, powerful and scalable, making them ideal to tackle large and highly complex Machine Learning tasks such as <strong>classifying billions of images(Google Images), powering speech recognition(Siri), recommending the best videos to watch to hundreds and millions of users(Youtube), or learning to beat the world champion at the game of Go(DeepMind's AlphaGo).</strong></mark></p>

<p style="font-size:120%">With that being said, let's have a look at the anatomy/structure of a biological neuron below.</p>

<h1 style="font-size:50px;color:#6166B3"><strong>Biological </strong><strong style="color:black">Neurons:</strong></h1>

![Neuron_Img](https://upload.wikimedia.org/wikipedia/commons/1/10/Blausen_0657_MultipolarNeuron.png)
<p style="font-size:120%;text-align:center">Figure 1: A Biological Neuron</p>

<p style="font-size:120%">Neurons are information messengers. <mark>They use <strong>electrical impulses and chemical signals</strong> to transmit information between different areas of the brain</mark>, and between the brain and the rest of the nervous system. Everything we think and feel and do would be impossible without the work of neurons and their support cells.</p>

<p style="font-size:120%"><mark>Neurons have three basic parts: a cell body and two extensions called an <strong>axon and a dendrite.</strong></mark> Within the cell body is a nucleus, which controls the cell’s activities and contains the cell’s genetic material. The axon looks like a long tail and transmits messages from the cell. Dendrites look like the branches of a tree and receive messages for the cell. <mark>Neurons communicate with each other <strong>by sending chemicals, called neurotransmitters, across a tiny space, called a synapse, between the axons and dendrites of adjacent neurons.</strong></mark></p>

<h1 style="font-size:50px;color:#6166B3"><strong>The First </strong><strong style="color:black">Simple Model:</strong></h1>

![ANN Computations](https://www.researchgate.net/profile/Bojan_Lukic2/publication/351547862/figure/fig2/AS:1022913150996480@1620892908389/ANNs-performing-simple-logical-computations-2.ppm)
<p style="font-size:120%;text-align:center">Figure 3: ANNs performing simple logical computations</p>

<p style="font-size:120%"><strong>Warren McCulloch and Walter Pitts</strong> proposed a very simple model of the biological neuron, which later became known as <strong>an artificial neuron</strong>:<mark> it has one or more binary(on/off) inputs and one binary output. The artificial neuron simply activates its out‐put when more than a certain number of its inputs are active.</mark> McCulloch and Pitts showed that even with such a simplified model it is possible to build a network of artificial neurons that computes any logical proposition you want.</p>

<p style="font-size:120%"><strong>Difficult to understand? Fear not.</strong> Let me explain:</p>
<ul>
    <li style="font-size:120%">Let's look at the first <strong>Artificial Neuron(C=A)</strong>. <mark>Here, only if there is a signal passed on by A to C, only then it will provide the output. Else there will not be any output.</mark> Therefore, here, C is dependent on the input passed on by A. Easy right? Let's see the next one.</li>
    <li style="font-size:120%">Next, if you see the second one, you will notice that, <mark>if signals are <strong>passed on to C by both A and B, then it will give an output.</strong></mark> Only A or B's signals by themselves are not alone to produce an output here.</li>
    <li style="font-size:120%"><strong>The (C=A OR B)</strong> <mark>will produce an output if C receives output from either A or B.</mark> Getting the hang of it? Let's look at a more complex example.</li>
    <li style="font-size:120%">In the final example, if we simplify the expression, it boils down to, <strong>A and NOT B</strong>. The way it works is as follows: <mark>If we receive <strong>some signal from A and B is off or the signal from B is not produced deliberately.</strong></mark> See! it's simple!</li>
</ul>
<p style="font-size:120%">These are some simple Artificial Neurons. They can be combined in various fashion to make more complex neural networks that can perform various tasks.</p>
