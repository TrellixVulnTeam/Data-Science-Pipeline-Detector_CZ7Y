```python
# Scaling the data
scaler = StandardScaler()

X1[:] = scaler.fit_transform(X1.values)
X2[:] = scaler.fit_transform(X2.values)
X3[:] = scaler.fit_transform(X3.values)
X4[:] = scaler.fit_transform(X4.values)
X5[:] = scaler.fit_transform(X5.values)
X6[:] = scaler.fit_transform(X6.values)
```
