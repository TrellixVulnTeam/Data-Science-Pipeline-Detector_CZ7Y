```python
from autoxgb import AutoXGB
```

```python
train_filename="../input/playground-nov-featured-data/new_train_1.csv"
output="./submition_2"
test_filename="../input/playground-nov-featured-data/new_test_1.csv"
idx="id"
targets=["target"]
use_gpu=True
num_folds=5
seed=42
num_trials=100
time_limit=3500
```

```python
axgb=AutoXGB(
    train_filename=train_filename,
    output=output,
    test_filename=test_filename,
    idx=idx,
    targets=targets,
    use_gpu=use_gpu,
    num_folds=num_folds,
    num_trials=num_trials,
    time_limit=time_limit
)
```

```python
axgb.train()
```
