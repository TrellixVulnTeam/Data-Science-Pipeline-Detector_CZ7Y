```python
import pandas as pd
import numpy as np
from autoxgb import AutoXGB
```

```python
!tree ../input/tabular-playground-series-nov-2021/
```

```python
train = pd.read_csv('../input/tabular-playground-series-nov-2021/train.csv')
test = pd.read_csv('../input/tabular-playground-series-nov-2021/test.csv')
sample_sub = pd.read_csv('../input/tabular-playground-series-nov-2021/sample_submission.csv')
```

```python
(train.shape, test.shape)
```

```python
train.head(10)
```

```python
train['target'].value_counts(normalize=True)
```

```python
axgb = AutoXGB(
    train_filename="../input/tabular-playground-series-nov-2021/train.csv",
    output="output_1",
    test_filename="../input/tabular-playground-series-nov-2021/test.csv",
    targets=['target'],
    use_gpu=True,
    task = None,
    idx = ['id'],
    num_folds=5,
    features = None,
    categorical_features = None,
    seed=42,
    num_trials=100,
    time_limit=360,
    fast=False,
)
axgb.train()
```
