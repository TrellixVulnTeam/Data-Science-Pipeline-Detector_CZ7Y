<center><h1 style = "font-size:20px;font-family: Copperplate">IMPORT THE LIBRARIES</h1></center>

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from tqdm import tqdm, trange
```

<center><h1 style = "font-size:20px;font-family: Copperplate">DATA LOADING</h1></center>

```python
train_data = pd.read_csv('/kaggle/input/tabular-playground-series-nov-2021/train.csv')
test_data = pd.read_csv('/kaggle/input/tabular-playground-series-nov-2021/test.csv')
```

```python
print(train_data.shape)
print(test_data.shape)

train_data.head()
```

<center><h1 style = "font-size:20px;font-family: Copperplate">Feature Statistics</h1></center>

```python
train_data.loc[:, 'f0':'f99'].describe().T.style.bar(subset=['mean'])\
                            .background_gradient(subset=['std'], cmap='Reds')\
                            .background_gradient(subset=['50%'], cmap='coolwarm')
```

<center><h1 style = "font-size:20px;font-family: Copperplate">Feature Distribution</h1></center>

```python
features = "f1 f3 f5 f8 f10 f17 f22 f25 f26 f34 f40 f41 f43 f50 f54 f55 f57 f66 f71 f80 f82 f91 f96 f97".split(' ')

fig, ax = plt.subplots(nrows=20, ncols=5, figsize=(20,100))
for i in trange(100):
    sns.kdeplot(train_data[train_data['target']==0][f'f{i}'], label='target=0', color='#DE3163', ax=ax[i//5, i%5])
    sns.kdeplot(train_data[train_data['target']==1][f'f{i}'], label='target=1', color='#58D68D', ax=ax[i//5, i%5])
    ax[i//5, i%5].set_xticks([])
    ax[i//5, i%5].set_xlabel(f'Feature {i}', fontsize=15)
    ax[i//5, i%5].axes.get_yaxis().set_visible(False)
    ax[i//5, i%5].spines['top'].set_visible(False)
    ax[i//5, i%5].spines['right'].set_visible(False)
    ax[i//5, i%5].spines['left'].set_visible(False)
    ax[i//5, i%5].spines['bottom'].set_visible(False)
    if f'f{i}' in features:
        ax[i//5, i%5].spines[:].set_visible(True)
        ax[i//5, i%5].spines[:].set_color('red')
        
ax[0,4].legend(fontsize=15)
plt.show()
```

- I highlighted the important features with red box.
- They have different distribution when grouped by targets. Therefore, they will be useful feature when modeling!
