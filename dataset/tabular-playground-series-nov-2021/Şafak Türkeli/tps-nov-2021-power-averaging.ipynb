```python
sample = pd.read_csv('../input/tabular-playground-series-nov-2021/sample_submission.csv')

# 0.74988, Simple NN-TPS Nov 21, https://www.kaggle.com/adityasharma01/simple-nn-tps-nov-21
sub1 = pd.read_csv('../input/simple-nn-tps-nov-21/submission.csv')
# 0.74987, TPS - Nov 2021 - NN - 2, https://www.kaggle.com/sfktrkl/tps-nov-2021-nn-2
sub2 = pd.read_csv('../input/tps-nov-2021-nn-2/submission.csv')
# 0.74966, TPS Nov 21 + DNN/CNN Model + Extras, https://www.kaggle.com/jmcslk/tps-nov-21-dnn-cnn-model-extras
sub3 = pd.read_csv('../input/tps-nov-21-dnn-cnn-model-extras/submission.csv')
# 0.74951, Simple NN with good results - TPS Nov 21, https://www.kaggle.com/javiervallejos/simple-nn-with-good-results-tps-nov-21
sub4 = pd.read_csv('../input/simple-nn-with-good-results-tps-nov-21/submission.csv')
# 0.74940, Understanding Neural Net, https://www.kaggle.com/chaudharypriyanshu/understanding-neural-net
sub5 = pd.read_csv('../input/understanding-neural-net/submission.csv')
```

# Check for correlations

```python
hist_data = [sub1.target, sub2.target, sub3.target, sub4.target, sub5.target]
group_labels = ['sub1', 'sub2', 'sub3', 'sub4', 'sub5']
fig = ff.create_distplot(hist_data, group_labels, bin_size=0.3, show_hist=False, show_rug=False)
fig.show()
```

```python
data = np.corrcoef([sub1.target, sub2.target, sub3.target, sub4.target, sub5.target])
fig=px.imshow(data,x=group_labels, y=group_labels)
fig.show()
```

High correlation between models 0.99+
> https://www.kaggle.com/c/tabular-playground-series-nov-2021/discussion/284507  
Note: Power Averaging to be used only when all the models are highly correlated, otherwise your score may become worse.

# Submission

```python
power = 4 # Power is arbitrary
ensemble = sample.copy()
ensemble.loc[:,'target'] = ((sub1**power + sub2**power + sub3**power + sub4**power + sub5**power) / 5) ** (1/power)
ensemble.to_csv('submission.csv', index=False)
ensemble
```
