```python
!pip3 install -U pandas==1.2.3 #Upgrading pandas
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import lazypredict
from lazypredict import Supervised
from lazypredict.Supervised import LazyClassifier


plt.style.use('fivethirtyeight')
plt.rcParams["figure.figsize"] = (20,5)
clear_output()
```

# <h1 style=' border:0; color:#FFA500 '><left>Loading Data üöö</left></h1> 

```python
import pandas as pd 
from sklearn.model_selection import train_test_split
import datetime

train = pd.read_csv("../input/tabular-playground-series-nov-2021/train.csv" )
test = pd.read_csv("../input/tabular-playground-series-nov-2021/test.csv")
```

```python
X = train.drop(['id','target'], axis=1)
y = train['target']
```

```python
#Spliting into training and validation set
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X,y,test_size = 0.20,random_state =42)
```

```python
del X,y,train,test
```

# <h1 style=' border:0; color:#FFA500 '><left>Model Selection üìö</left></h1> 

```python
num_models = 29 #Total number of model is 29 for classification models
selected_num_models = 22
reg_idx = [i for i in range(selected_num_models)]
noregs_idx = [4,9,9,9,13,20,20] # Removing 7 models from 29 models. Some of these models are time consuming whereas other require lot of ram and some getting error
regs_name =[]
regs = []
for i in range(num_models):
    regs_name.append(lazypredict.Supervised.CLASSIFIERS[i][0])
    regs.append(lazypredict.Supervised.CLASSIFIERS[i][1])

for i in noregs_idx:
    del regs_name[i]
    del regs[i]
```

```python
print("ALL 22 AVAILABLE CLASSIFICATION MODELS:")
for i in range(selected_num_models):
    print(i+1 , regs_name[i])
```

# <h1 style=' border:0; color:#FFA500 '><left>Training on 22 different models üèãÔ∏è</left></h1> 

```python
results = pd.DataFrame()
for i in range(selected_num_models):
    print(i,regs_name[i])
    reg = LazyClassifier(verbose=0, 
                    ignore_warnings=False,
                    custom_metric=None,
                    classifiers = [regs[i]])
    models, predictions = reg.fit(X_train, X_valid, y_train, y_valid)
    models.index = [regs_name[i]]
    results = results.append(models)
    print(results)
clear_output()
```

# <h1 style=' border:0; color:#FFA500 '><left>‚úîÔ∏è Results</left></h1> 

```python
print(results)
```

# <h1 style=' border:0; color:#FFA500 '><left> Comparison Plots üìà </left></h1> 

```python
plt.plot(reg_idx , results["ROC AUC"],label = "ROC AUC" ,marker='o')
plt.xlabel("Model ID")
plt.ylabel("ROC AUC")
plt.title("ROC AUC Comparison of 22 Different Models")
plt.legend()
plt.show()
```

```python
plt.plot(reg_idx , results["Time Taken"],label = "Time Taken" , color = 'r')
plt.xlabel("Model ID")
plt.ylabel("Time Taken")
plt.title("TIME TAKEN for 22 Different Models")
plt.legend()
plt.show()
```

# <h1 style=' border:0; color:#FFA500 '><left>ü•á Top 5 Performing Models</left></h1> 

```python
results = results.sort_values(by = "ROC AUC",ascending=False)
```
