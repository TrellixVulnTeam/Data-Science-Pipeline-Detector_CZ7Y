```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns
```

```python
train_df = pd.read_csv('../input/tabular-playground-series-nov-2021/train.csv',index_col='id')
test_df = pd.read_csv('../input/tabular-playground-series-nov-2021/test.csv',index_col='id')
submission_df = pd.read_csv('../input/tabular-playground-series-nov-2021/sample_submission.csv')
```

```python
print(f'Number of rows in train data is {train_df.shape[0]}. Number of columns in train data is {train_df.shape[1]}')
print(f'Number of rows in test data is {test_df.shape[0]}. Number of columns in test data is {test_df.shape[1]}')
```

```python
print(train_df.info())
```

```python
print(train_df.select_dtypes('float64').columns)
print(train_df.select_dtypes('int64').columns)
```

100 float (f0-f99) columns and one int (target)

```python
features = train_df.select_dtypes('float64').columns
target = train_df.select_dtypes('int64').columns
```

```python
train_df.isnull().sum().sum()
```

No null values.

```python
#Code Copied from one of noteooks of @Subin An

train_df[features].describe().T.style.bar(subset='mean',color='deepskyblue')\
                                .background_gradient(subset='std',cmap='Greens')
```

Right skewed distributions. Scaling will be needed as  f2 ,f35 are on different scale.

```python
train_df[features].duplicated().sum()
```

No duplicates

```python
#Code Copied from one of noteooks of @Subin An


fig, axes = plt.subplots(10,10,figsize=(20,20))

for col,ax in zip(features,axes.flatten()):
    sns.kdeplot(x=col,data=train_df,fill=True,ax=ax,color='Deepskyblue',label='Train')
    sns.kdeplot(x=col,data=test_df,fill=True,ax=ax,color='Green',label='Test')
    
    if col == 'f9':
        ax.legend()
    
    ax.spines.top.set_visible(False)
    ax.spines.right.set_visible(False)
    ax.spines.left.set_visible(False)
    
    ax.set_xticklabels([])
    ax.set_xticks([])
    ax.set_yticklabels([])
    ax.set_yticks([])
    ax.set_ylabel('')


plt.suptitle('Distribution of continuous features (Train and Test)',fontweight = 'bold')

plt.tight_layout()
plt.show()
```

Test and Train Data have similar distribution (expected).

```python
fig,ax = plt.subplots(1,1,figsize=(10,10))

sns.countplot(x='target',data=train_df,ax=ax)
ax.set_title("Target Distribution",fontweight='bold')
plt.show()
```

No Class Imbalance

```python
fig,ax = plt.subplots(1,1,figsize=(20,20))

corr = train_df.corr()

mask = np.triu(np.ones_like(corr,dtype='bool'))

sns.heatmap(corr,mask=mask,ax=ax)

ax.set_title('Correlation among features and target',fontweight='bold')

plt.show()
```

Correlation among features and target is low
