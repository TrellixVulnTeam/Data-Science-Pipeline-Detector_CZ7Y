```python
train_data = pd.read_csv("/kaggle/input/tabular-playground-series-nov-2021/train.csv")
test_data = pd.read_csv("/kaggle/input/tabular-playground-series-nov-2021/test.csv")
submission_data = pd.read_csv("/kaggle/input/tabular-playground-series-nov-2021/sample_submission.csv")
```

```python
train_data.head()
train_data.describe()
```

```python
# y=train_data.target
#columns=train_data.columns[1:101]
columns=["f1","f3","f5","f6","f7","f8","f10","f11","f13"]
X=train_data[columns]
X.head()
y=train_data.target
```

```python
#test_data.head()
test_input=test_data[columns]
test_input.head()
```

```python
from xgboost import XGBRegressor
model = XGBRegressor()

model.fit(X, y)

# Generate test predictions
predict = model.predict(test_input)

```

```python
my_submission = pd.DataFrame({'id': test_data.id, 'target': predict})
# you could use any filename. We choose submission here
my_submission.to_csv('submission.csv', index=False)
my_submission.head()
```
