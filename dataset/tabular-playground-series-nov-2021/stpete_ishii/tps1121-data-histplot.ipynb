```python
cols=train.columns.tolist()
print(cols)
```

```python
fig, ax = plt.subplots(25,4,figsize=(16,80))
for i in tqdm(range(100)):
    if i<100:
        r=i//4
        c=i%4
        sns.histplot(train[train.target==0][cols[i]], label=cols[i]+' target=0', ax=ax[r,c], color='black',bins=20)
        sns.histplot(train[train.target==1][cols[i]], label=cols[i]+' target=1', ax=ax[r,c], color='C1',bins=20)
        ax[r,c].legend()
        ax[r,c].grid()
    else:
        r=i//4
        c=i%4
        ax[r,c].axis("off")       
plt.show()
```
