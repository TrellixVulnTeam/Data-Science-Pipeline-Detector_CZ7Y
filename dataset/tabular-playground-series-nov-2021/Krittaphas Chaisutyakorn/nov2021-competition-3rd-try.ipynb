```python
train = pd.read_csv("../input/tabular-playground-series-nov-2021/train.csv")
```

```python
train_id = train[['id']]
train_target = train[['target']]
train_data = train.drop(['id', 'target'], axis=1)
```

```python
train_data.head()
```

```python
train_data.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(train_data, train_target, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

log_clf = Pipeline([
    ('scaler', StandardScaler()),
    ('log', LogisticRegression()),
])
```

```python
log_clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import roc_auc_score

y_prob_train = log_clf.predict_proba(X_train)

auc_score = roc_auc_score(y_train, y_prob_train[: ,1])
auc_score
```

```python
y_prob_test = log_clf.predict_proba(X_test)

auc_score_test = roc_auc_score(y_test, y_prob_test[: ,1])
auc_score_test
```

```python
# finalize the model 
log_clf.fit(train_data, train_target)
```

```python
test = pd.read_csv("../input/tabular-playground-series-nov-2021/test.csv")
```

```python
test_id = test[['id']]
test_data = test.drop('id', axis=1)
```

```python
test_data
```

```python
y_prob_test = log_clf.predict_proba(test_data)
```

```python
score = y_prob_test[:, 1]
score
```

```python
submission_lr = test_id
submission_lr
```

```python
submission_lr['target'] = score
```

```python
submission_lr = submission_lr.set_index('id')
submission_lr
```
