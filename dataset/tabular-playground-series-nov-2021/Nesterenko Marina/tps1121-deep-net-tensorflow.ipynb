```python
import pandas as pd
from sklearn.cluster import KMeans
```

```python
dataset_train1 = pd.read_csv('../input/tabular-playground-series-nov-2021/train.csv', index_col='id')
dataset_test1 = pd.read_csv('../input/tabular-playground-series-nov-2021/test.csv', index_col='id')
y = dataset_train1.target
y = pd.DataFrame(y)
dataset_train = dataset_train1.drop(['target'], axis=1)
```

```python
for col in dataset_train.columns:
    dataset_train[col] = dataset_train[col].astype('float32')
    dataset_test1[col] = dataset_test1[col].astype('float32')
dataset_train.dtypes
```

```python
model_kmeans = KMeans(n_clusters=5, random_state=59)
model_kmeans.fit(dataset_train);

clusters_train = model_kmeans.predict(dataset_train)
clusters_test  = model_kmeans.predict(dataset_test1)

dataset_train['fe_cluster'] = clusters_train
dataset_test1['fe_cluster']  = clusters_test
```

```python
dataset_train = pd.get_dummies(dataset_train, columns=['fe_cluster'])
dataset_test1 = pd.get_dummies(dataset_test1, columns=['fe_cluster'])
```

```python
dataset_train['fe_mean']        = dataset_train.mean(axis=1)
dataset_test1['fe_mean']         = dataset_test1.mean(axis=1)
dataset_train['fe_median']      = dataset_train.median(axis=1)
dataset_test1['fe_median']       = dataset_test1.median(axis=1)
dataset_train['fe_min']         = dataset_train.min(axis=1)
dataset_test1['fe_min']          = dataset_test1.min(axis=1)
dataset_train['fe_max']         = dataset_train.max(axis=1)
dataset_test1['fe_max']          = dataset_test1.max(axis=1)
dataset_train['fe_skew']        = dataset_train.skew(axis=1)
dataset_test1['fe_skew']         = dataset_test1.skew(axis=1)
dataset_train['sum1']        = dataset_train.sum(axis=1)
dataset_test1['sum1']         = dataset_test1.sum(axis=1)
```

```python
dataset_train['f0_f1'] = dataset_train['f0'] + dataset_train['f1']
dataset_test1['f0_f1'] = dataset_test1['f0'] + dataset_test1['f1']
dataset_train['f1_f2'] = dataset_train['f2'] + dataset_train['f1']
dataset_test1['f1_f2'] = dataset_test1['f2'] + dataset_test1['f1']
```

```python
for col in dataset_train.columns:
    dataset_train[f'{col}^2'] = dataset_train[col]**2
    dataset_test1[f'{col}^2'] = dataset_test1[col]**2
```

```python
from tensorflow import keras
import tensorflow as tf
from tensorflow.keras.layers import Dense, Embedding
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import Reshape
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Conv1D
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.models import load_model
#from tensorflow.keras.engine.input_layer import Input
from tensorflow.keras.layers import MaxPooling1D, GaussianNoise
from tensorflow.keras.layers import GlobalAveragePooling1D
from tensorflow.keras.layers import BatchNormalization
```

```python
from tensorflow.keras.constraints import max_norm
```

```python
shape = len(dataset_train_sc[0])
```

```python
def model_builder(lr):
    """Модель нейронной сети для агента"""
    inputA = keras.Input(shape=(shape))
    #line = GaussianNoise(0.00001)(inputA)
    line = Reshape((shape,1))(inputA)
    
    line = Conv1D(filters=32, kernel_size=2, padding='same', activation='swish')(line)
    line = BatchNormalization()(line)
    line = Dropout(0.4)(line)
    
    line = Conv1D(filters=64, kernel_size=2, activation='swish')(line)
    line = BatchNormalization()(line)
    line = Dropout(0.4)(line)
    
    line = Conv1D(filters=128, kernel_size=2, activation='swish')(line)
    line = BatchNormalization()(line)
    line = Dropout(0.4)(line)
    
    #line = MaxPooling1D(pool_size=2)(line)
    
    line = Flatten()(line)
    #line = Dense(512, activation='relu', kernel_regularizer=keras.regularizers.l1(0.000001))(line)
    #line = Dense(256, activation='relu', kernel_regularizer=keras.regularizers.l1(0.000001))(line)
    #line = Dropout(0.1)(line)
    #line = Dense(128, activation='relu')(line)
    #line = Dense(32, activation='relu')(line)
    #line = Dense(32, activation='relu',kernel_constraint=max_norm(2.),)(line)
    #line = Dense(256, activation='swish')(inputA)
    #line = Dropout(0.5)(line)
    #line = Dense(128, activation='swish')(line)
    #line = Dropout(0.5)(line)
    line = Dense(256, activation='swish')(line)
    line = Dropout(0.4)(line)
    line = Dense(128, activation='swish')(line)
    line = Dropout(0.4)(line)
    line = Dense(64, activation='swish')(line)
    line = Dropout(0.4)(line)
    line = Dense(32, activation='swish')(line)
    
    #line = Dense(16, activation='relu')(line)
    
    outputA = Dense(units=1, activation="sigmoid", kernel_regularizer=keras.regularizers.l1(0.01))(line)
    model = Model(inputs=inputA, outputs=outputA)
    #model = keras.models.load_model('models/model2')
    model.compile(loss = 'binary_crossentropy', optimizer = Adam(lr=lr), metrics=[tf.keras.metrics.AUC(name='auc'), 'binary_accuracy'],)
    return model
```

```python
lr=0.0001
with strategy.scope():
    model = model_builder(lr)
```
