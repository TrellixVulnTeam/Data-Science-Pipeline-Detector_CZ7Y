# import libraries

```python
import numpy as np
import pandas as pd 
import h2o
from h2o.automl import H2OAutoML
```

# define variable

```python
TRAIN_PATH = "../input/tabular-playground-series-nov-2021/train.csv"
TEST_PATH = "../input/tabular-playground-series-nov-2021/test.csv"
SAMPLE_SUBMISSION_PATH = "../input/tabular-playground-series-nov-2021/sample_submission.csv"
SUBMISSION_PATH = "submission.csv"

TARGET = "target"
ID = "id"
SEED = 2021
MAX_RUNTIME_SECS = 60 * 20
```

# init h20

```python
h2o.init()
```

# load data

```python
# load data
train = h2o.import_file(TRAIN_PATH)
test = h2o.import_file(TEST_PATH)
```

# split data

```python
x = train.columns
y = TARGET

x.remove(y)
x.remove(ID) #remove id  
```

# build model

```python
aml_y = H2OAutoML(max_runtime_secs=MAX_RUNTIME_SECS, seed=SEED)
aml_y.train(x=x, y=y, training_frame=train)
```

# predict test data

```python
preds_y = aml_y.predict(test)
```

# submit prediction result

```python
submission = pd.read_csv(SAMPLE_SUBMISSION_PATH)
submission[TARGET] = preds_y.as_data_frame().predict
submission.to_csv(SUBMISSION_PATH, index=False)
```
