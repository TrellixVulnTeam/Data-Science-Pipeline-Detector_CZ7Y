```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

from sklearn.model_selection import train_test_split

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams["figure.figsize"] = (5,10)

import os
import copy
import gc
import pickle
```

Read dataset.

```python
df_train = pd.read_csv("../input/tabular-playground-series-nov-2021/train.csv")
df_test = pd.read_csv("../input/tabular-playground-series-nov-2021/test.csv")
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
y = df_train.pop("target")
```

```python
FEATURES = df_train.columns[1:]
FEATURES
```

No feature has nan values. So we normalize.

```python
for f in FEATURES:
  df_train[f] = (df_train[f] - df_train[f].min())
  df_train[f] = df_train[f] / df_train[f].max()
  df_test[f] = (df_test[f] - df_test[f].min())
  df_test[f] = df_test[f] / df_test[f].max()
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_train[FEATURES].to_numpy(), y.to_numpy(), test_size=0.2, shuffle=True)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
model = keras.models.Sequential()
model.add(layers.Dense(128, input_shape=(100, ), activation="relu"))
model.add(layers.Dense(128, activation="relu"))
model.add(layers.Dense(128, activation="relu"))
model.add(layers.Dense(128, activation="relu"))
model.add(layers.Dense(128, activation="relu"))
model.add(layers.Dense(1, activation="sigmoid"))

model.compile(loss=keras.losses.MeanSquaredError(), optimizer="adam", metrics=["AUC"])
model.summary()
```

```python
model.fit(X_train, y_train, validation_data=(X_test, y_test), batch_size=64, epochs=10)
```

```python
pred = model.predict(df_test[FEATURES].to_numpy())
```

```python
pred
```

```python
df_pred = pd.DataFrame(pred, index=df_test["id"], columns=["target"])
```

```python
df_pred.to_csv(os.path.join("sub.csv"), index=True)
```
