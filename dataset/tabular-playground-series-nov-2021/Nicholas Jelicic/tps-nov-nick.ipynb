```python
df_train = pd.read_csv('/kaggle/input/tabular-playground-series-nov-2021/train.csv')
df_test = pd.read_csv('/kaggle/input/tabular-playground-series-nov-2021/test.csv')
```

```python
#target kan [0,1] -> classificatie

df_train.head()

df_train = df_train
```

```python
df_test # nog geen target, deze gaan we voorspellen obv het model dat we gaan maken met df_train
```

```python
X = df_train[df_train.columns[1:-1]]
y = df_train[df_train.columns[-1]]

clf = LogisticRegression(penalty='none',solver='sag')

clf.fit(X,y)
```

```python
predictions = clf.predict_proba(df_test[df_test.columns[1:]])
```

```python
df_sub = pd.read_csv('/kaggle/input/tabular-playground-series-nov-2021/sample_submission.csv')

df_sub['target'] = predictions
df_sub
```

```python
df_sub.to_csv('submission.csv',index=False)
```
