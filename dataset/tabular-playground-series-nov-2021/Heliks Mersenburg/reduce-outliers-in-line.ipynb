# reduces outliers to limit value using boxplot resources

```python
def boxplott(df,text):
    Q1=df[text].quantile(q=0.25)
    Q3=df[text].quantile(q=0.75)
    FIQ = Q3-Q1

    LF = Q1 - 1.5 * FIQ
    LS = Q3 + 1.5 * FIQ
    df.loc[df[text] > LS , text] = LS
    df.loc[df[text] < LF , text] = LF
    return df
```

# Graph plot of any line
*  Let's convert any row into a single-column dataframe and plot the graph.

```python
def plot_graph_alone(df,val):
    df_index=df.loc[df['id'] == val]
    fmri =[]
    count=0
    df_index=df_index.to_numpy()
    for value in df_index[0]:
        if count < 1 or count > 90:
            count=+1
        else:
            fmri.append(np.deg2rad(value))
            count=+1
    fmri=pd.DataFrame(fmri)
    fmri = fmri[:-1]
    fmri.columns=['value']
    fmri = fmri.rename_axis('index').reset_index()
    
    fmri = fmri.melt('index', var_name='cols',  value_name='vals')
    
    g = sns.lineplot(x="index", y="vals", hue='cols', data=fmri) 
plot_graph_alone(df,5341)
```

# 5 Random Line Graph Plot

* The function transforms 5 rows into columns and plots the comparative graph.

*  Note that there are outliers for each series.

```python
def plot_graph_random(df):
    max_iter = 5
    seed(max_iter)
    df_fmri=pd.DataFrame()
    rand_list=[]
    for index in range(max_iter):
        rand=randint(0, 6000)
        rand_list.append(rand)
        df_index=df.loc[df['id'] == rand]
        fmri =[]
        count=0
        df_index=df_index.to_numpy()
        for value in df_index[0]:
            if count < 1 or count > 90:
                count=+1
            else:
                fmri.append(value)
                count=+1
        df_fmri[rand]=fmri
    df_fmri = df_fmri[:-1]
        
    df_fmri = df_fmri.rename_axis('index').reset_index()
    
    df_fmri = df_fmri.melt('index', var_name='cols',  value_name='vals')
    
    g = sns.lineplot(x="index", y="vals", hue='cols', data=df_fmri) 
plot_graph_random(df)
```

# Plotting 5 random lines, reducing outliers to the lower and upper limits

* Outlier values make it difficult to observe smaller values, for that we will reduce them to the respective lower and upper limit values using boxplot.

```python
def plot_graph_reduces(df):
    max_iter = 5
    seed(max_iter)
    df_fmri=pd.DataFrame()
    rand_list=[]
    for index in range(max_iter):
        rand=randint(0, 6000)
        rand_list.append(rand)
        df_index=df.loc[df['id'] == rand]
        fmri =[]
        count=0
        df_index=df_index.to_numpy()
        for value in df_index[0]:
            if count < 1 or count > 90:
                count=+1
            else:
                fmri.append(value)
                count=+1
        df_fmri[rand]=fmri
    df_fmri = df_fmri[:-1]
    
    for i in rand_list:
        df_fmri= boxplott(df_fmri,i)
    
    df_fmri = df_fmri.rename_axis('index').reset_index()
    
    df_fmri = df_fmri.melt('index', var_name='cols',  value_name='vals')
    
    g = sns.lineplot(x="index", y="vals", hue='cols', data=df_fmri) 
plot_graph_reduces(df)
```
