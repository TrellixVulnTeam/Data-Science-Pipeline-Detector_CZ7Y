```python
from autoxgb import AutoXGB
```

```python
# path to training data
train_filename = "../input/tabular-playground-series-nov-2021/train.csv"

# path to output folder to store artifacts
output = "fayadh"
```

```python
test_filename = "../input/tabular-playground-series-nov-2021/test.csv"
```

```python
task = "classification"
idx =None
num_folds = 10
features=None
categorical_features=None
targets = ["target"]
use_gpu = True
seed = 1234
num_trials = 2000
time_limit = 20000
fast = False
```

```python
axgb = AutoXGB(
    train_filename=train_filename,
    output=output,
    test_filename=test_filename,
    task=task,
    idx=idx,
    targets=targets,
    features=features,
    categorical_features=categorical_features,
    use_gpu=use_gpu,
    num_folds=num_folds,
    seed=seed,
    num_trials=num_trials,
    time_limit=time_limit,
    fast=fast,
)
axgb.train()
```

```python
import pandas as pd
test_preds = pd.read_csv("/kaggle/working/fayadh/test_predictions.csv")
sample_submission = pd.read_csv("../input/tabular-playground-series-nov-2021/sample_submission.csv")
sample_submission.target = test_preds["1"].values
sample_submission.to_csv("submission.csv", index=False)
```
