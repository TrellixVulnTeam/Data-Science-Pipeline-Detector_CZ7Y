```python
import tensorflow as tf
from tensorflow.keras.layers import Dense, BatchNormalization, Dropout
from tensorflow.keras import Sequential
```

```python
os.getcwd()
```

```python
train_data = pd.read_csv('../input/tabular-playground-series-nov-2021/train.csv')
train_data.head()
```

```python
test_data = pd.read_csv('../input/tabular-playground-series-nov-2021/test.csv')
```

```python
data_train = train_data.copy()
data_test = test_data.copy()
```

```python
data_train.head()
```

```python
for i in data_train.columns:
    print('\'{}\''.format(i), end = ', ')
print('\b\b')
```

```python
x = data_train.loc[:, 'f0': 'f99']
```

```python
x
```

```python
y = data_train['target']
y
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
x_scaler = StandardScaler()
x_scaler.fit(x)
```

```python
x = x_scaler.transform(x)
x
```

```python
input_size = x[0].shape

model = Sequential([
    Dense(64, activation = 'relu', input_shape = input_size),
    Dense(64, activation = 'relu'),
    Dense(128, activation = 'relu'),
    Dense(128, activation = 'relu'),
    Dense(64, activation = 'relu'),
    Dense(32, activation = 'relu'),
    Dense(32, activation = 'relu'),
    Dense(1, activation = 'sigmoid')
])

model.compile(optimizer = tf.keras.optimizers.RMSprop(learning_rate = 0.001), loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
history = model.fit(x, y, batch_size = 256, validation_split = 0.2, epochs = 10, verbose = 1)
hist  = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
```

```python
import matplotlib.pyplot as plt

plt.plot(hist['loss'])
plt.plot(hist['val_loss'],)
plt.legend(("train" , "valid") , loc =0)
```
