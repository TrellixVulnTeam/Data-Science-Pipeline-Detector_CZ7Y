```python
## import packages
from evalml.automl import AutoMLSearch
import pandas as pd
```

```python
train = pd.read_csv('../input/tabular-playground-series-nov-2021/train.csv')
test = pd.read_csv('../input/tabular-playground-series-nov-2021/test.csv')
sub = pd.read_csv('../input/tabular-playground-series-nov-2021/sample_submission.csv')
train.shape, test.shape, sub.shape
```

```python
target=train['target']
train.drop(['target'],1,inplace=True)
```

```python
## run model
model_evalml = AutoMLSearch(X_train=train, y_train=target, problem_type='binary', max_time=3600)
model_evalml.search()
```

```python
## check leaderboard
model_evalml.rankings
```

```python
## generate predictions
preds_evalml = model_evalml.best_pipeline.predict_proba(test)[True]
```

```python
## create submission
import pandas as pd
submission=pd.DataFrame()
submission['id'] = test['id']
submission['target']=preds_evalml
```
