# Time to Run
<div align="center">
  <img width="900"  src="https://i.postimg.cc/6QDzDvhT/Copy-of-Copy-of-Add-a-heading.gif">
</div>

```python
import timeit

time_pp = timeit.timeit(lambda: ProfileReport(data),number=1)
time_av = timeit.timeit(lambda: AV.AutoViz('', dfte = data),number=1)
time_dp = timeit.timeit(lambda: create_report(data), number=1)
time_sv = timeit.timeit(lambda: sv.analyze(data), number=1)
time_dt = 1.47 # Calculated 
```

```python
time = {'Package' :['Pandas Profiling', 'Autoviz', 'Dataprep', 'SweetViz', 'D-Tale'],
           'Time' : [time_pp, time_av, time_dp, time_sv, time_dt]}

time = pd.DataFrame(time).sort_values('Time', ignore_index=True)
time.Time = time.Time.round(2)

fig = px.line(time,
            x = time.Package,
            y = time.Time,
            text=time.Time,
        )
fig.update_traces(textposition = 'bottom right')
fig.update_layout(title = 'Time(seconds) taken by Packages',
                  plot_bgcolor = '#e4d1eb')

fig.show()
```
