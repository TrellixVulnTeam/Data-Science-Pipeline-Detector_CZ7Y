```python
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
train_data = TabularDataset('/kaggle/input/tabular-playground-series-nov-2021/train.csv').drop('id', axis=1)
train_data.head()
```

```python
test_data = TabularDataset('../input/tabular-playground-series-nov-2021/test.csv').drop('id', axis=1)
test_data.head()
```

```python
metric = 'roc_auc'
time=3*3600
predictor = TabularPredictor(label='target', eval_metric=metric).fit(
    train_data= train_data, time_limit=time, 
    ag_args_fit={'num_gpus': 1},
    presets='best_quality', verbosity=2)
predictor.leaderboard(train_data, silent=True)
```

```python
results = predictor.fit_summary()
```

```python
pred_probs = predictor.predict_proba(test_data)
pred_probs.head(5)
```

```python
submission = pd.read_csv('../input/tabular-playground-series-nov-2021/sample_submission.csv')
submission.target = pred_probs[1]
submission.to_csv('submission.csv', index=False)
```
