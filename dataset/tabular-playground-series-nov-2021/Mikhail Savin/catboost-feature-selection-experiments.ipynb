## CatBoostClassifier.select_features()

In pursuit for ROC increase i tried to make several test with batch of features with CatBoost select_features tool
In some cases its give me ROC increase on validation dataset up to ~ 0.02

While waiting submition time reset wanna share with you, maybe somebody will find the best of the best features combination :)

```python
import pandas as pd
import numpy as np

from catboost import CatBoostClassifier, Pool
```

```python
train = pd.read_csv('../input/tabular-playground-series-nov-2021/train.csv').drop(['id'], axis=1)

train_data = train.drop(['target'], axis=1)
train_labels = train['target']
```

```python
model = CatBoostClassifier(
    iterations=20,
    learning_rate=0.05,
    l2_leaf_reg=3,
    depth=8,
    
)

train_pool = Pool(train_data, 
                  train_labels)
```

```python
result = model.select_features(
    train_pool,
    features_for_select=train_data.columns.tolist(),
    num_features_to_select=70
)
```

```python
selected_features = result['selected_features_names']
selected_features
```
