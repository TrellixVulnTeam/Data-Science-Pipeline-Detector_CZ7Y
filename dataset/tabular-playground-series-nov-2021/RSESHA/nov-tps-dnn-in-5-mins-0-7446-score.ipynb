```python
from deep_autoviml import deep_autoviml as deepauto
```

```python
keras_options = {}
model_options = {}
```

```python
# define train and test files here
train_filename="../input/tps-nov-features/train_features.csv"
target='target'
keras_model_type = "fast1"
project_name = "TPS_Nov21"
```

# There is a minor problem with train and test: there is one extra feature in Train (kfold) than in Test. Hence we need to remove that feature and train the Deep Learning model.

```python
import pandas as pd
train = pd.read_csv(train_filename).drop('kfold', axis=1)
print(train.shape)
train.head(1)
```

```python

# deep_autoviml can use multiple model types - we will use the fas1 which is a deep and wide network
model, cat_vocab_dict = deepauto.fit(train, target, keras_model_type=keras_model_type,
		project_name=project_name, keras_options=keras_options,  
		model_options=model_options, save_model_flag=True, use_my_model='',
		model_use_case='', verbose=1)
```

# It took less than 5 mins to build a deep and wide model with balanced accuracy on validation data set of 71%. Let's see how well the model does on test data.

```python
test_filename="../input/tps-nov-features/test_features.csv"
test = pd.read_csv(test_filename)
print(test.shape)
test.head(1)
```

```python
predictions = deepauto.predict(model, project_name, test_dataset=test,
                                 keras_model_type=keras_model_type, 
                                 cat_vocab_dict=cat_vocab_dict)
len(predictions)
```

```python
predictions[0]
```

```python
import pandas as pd
sample_submission = pd.read_csv("../input/tabular-playground-series-nov-2021/sample_submission.csv")
sample_submission.target = predictions[0][:,1]
sample_submission.to_csv("submission.csv", index=False)
```
