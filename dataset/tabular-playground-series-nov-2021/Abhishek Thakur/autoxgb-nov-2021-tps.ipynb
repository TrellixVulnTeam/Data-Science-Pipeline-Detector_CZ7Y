```python
from autoxgb import AutoXGB
```

```python
# AutoXGB will use kfold column, if available, for cross-validation folds
axgb = AutoXGB(
    train_filename="../input/tps-nov-features/train_features.csv",
    output="tps",
    test_filename="../input/tps-nov-features/test_features.csv",
    targets=['target'],
    idx=None,
    use_gpu=True,
    task="classification",
    features=None,
    categorical_features=None,
    seed=42,
    num_trials=1000,
    time_limit=25000,
    fast=False,
)
axgb.train()
```

```python
import pandas as pd
test_preds = pd.read_csv("/kaggle/working/tps/test_predictions.csv")
sample_submission = pd.read_csv("../input/tabular-playground-series-nov-2021/sample_submission.csv")
sample_submission.target = test_preds["1"].values
sample_submission.to_csv("submission.csv", index=False)
```
