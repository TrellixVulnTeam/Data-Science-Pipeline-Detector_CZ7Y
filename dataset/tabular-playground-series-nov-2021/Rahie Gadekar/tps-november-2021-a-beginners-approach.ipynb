```python
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense , Flatten, BatchNormalization ,Dropout, Input, MaxPooling2D
from tensorflow.keras import layers
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint , ReduceLROnPlateau
from tensorflow import keras
```

```python
from sklearn.preprocessing import StandardScaler, RobustScaler
from sklearn.model_selection import train_test_split 

x = df.drop(["id" , "target"], axis = 1 )
y = df['target']

sc = RobustScaler()
x = pd.DataFrame(columns = x.columns , data = sc.fit_transform(x))

xtrain, xtest , ytrain , ytest = train_test_split(x , y , test_size = 0.2 , random_state = 0)
```

```python
from tensorflow import keras
def model_builder(hp):
    model = keras.Sequential()
    # Tune the number of units in the first Dense layer
    # Choose an optimal value between 32-512
    hp_units = hp.Int('units', min_value=32, max_value=512, step=32)
    model.add(keras.layers.Dense(units=hp_units, activation='swish'))
    model.add(keras.layers.Dense(64))
    model.add(keras.layers.Dense(32))
    model.add(keras.layers.Dense(16))
    model.add(BatchNormalization())
    model.add(keras.layers.Dense(1 , activation = 'sigmoid'))

    # Tune the learning rate for the optimizer
    # Choose an optimal value from 0.01, 0.001, or 0.0001
    hp_learning_rate = hp.Choice('learning_rate', values=[1e-2, 1e-3, 1e-4])
    
    model.compile(optimizer=keras.optimizers.Adam(learning_rate=hp_learning_rate),
                loss=keras.losses.BinaryCrossentropy(),
                metrics=['AUC'])

    return model
```

```python
def createmodel() : 
    model = Sequential()
    model.add(Dense(units = 224 , activation = 'swish' , input_dim = 100))
    model.add(Dense(units = 224 , activation = 'swish'))
    model.add(Dense(units = 224 , activation = 'swish'))
    model.add(Dropout(0.5))
    model.add(Dense(units = 128 , activation = 'swish'))
    model.add(Dense(units = 128 , activation = 'swish'))
    model.add(Dense(units = 128 , activation = 'swish'))
    model.add(Dense(units = 128 , activation = 'swish'))
    model.add(Dropout(0.5))
    model.add(Dense(units = 64 , activation = 'swish'))
    model.add(Dense(units = 64 , activation = 'swish'))
    model.add(Dense(units = 64 , activation = 'swish'))
    model.add(Dense(units = 64 , activation = 'swish'))
    model.add(Dropout(0.5))
    model.add(Dense(units = 64 , activation = 'swish'))
    model.add(Dense(units = 64 , activation = 'swish'))
    model.add(Dense(units = 64 , activation = 'swish'))
    model.add(Dense(units = 64 , activation = 'swish'))
    model.add(Dropout(0.5))
    model.add(Dense(units = 32 , activation = 'swish'))
    model.add(Dense(units = 32 , activation = 'swish'))
    model.add(BatchNormalization())
    model.add(Dense(units = 1 , activation = 'sigmoid'))
    model.compile(optimizer = 'adam' , loss = 'binary_crossentropy' , metrics = ['AUC'])
    return model 
```

```python
best_model = createmodel()
```

```python
es = EarlyStopping(patience = 10)
ch = ModelCheckpoint(filepath = "model1.h5",save_best_only = True)
lr = ReduceLROnPlateau(monitor='val_loss')
```

```python
from sklearn.model_selection import StratifiedKFold
Batch_size = 500 
epochs = 10
sf = StratifiedKFold(n_splits = 10 , shuffle = True , random_state = 0)
scores = {fold : None for fold in range(sf.n_splits)}
predictions = []
for fold, (train_idx , test_idx) in enumerate(sf.split(x , y)) : 
    xtrain , ytrain = x.iloc[train_idx] , y.iloc[train_idx]
    xtest , ytest = x.iloc[test_idx] , y.iloc[test_idx] 
    
 #   model = createmodel()
    print("**"*20)
    print(f"fold : {fold +1} || Training")
    print("**"*20)
    model_history = best_model.fit(xtrain , ytrain , batch_size = Batch_size , epochs = epochs, callbacks = [es ,lr,  ch], validation_data=(xtest, ytest))
    scores[fold] = (model_history.history)
    print(f"fold : {fold + 1} || Validation AUC : {np.max(scores[fold]['val_auc'])}")
    
    prediction = best_model.predict(xtest , batch_size = Batch_size).reshape(1 , -1)[0]
    predictions.append(prediction)

print("**"*20)
print("Finished Training")
print("**"*20)

overall_auc = [np.max(scores[fold]['val_auc']) for fold in range(sf.n_splits)]
print("Overall AUC : ", np.mean(overall_auc))
```
