# 1. Import required library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# 2. Import Dataset with Tensorflow

```python
from tensorflow.keras.datasets import mnist
```

```python
(x_train, y_train),(x_test, y_test) = mnist.load_data()
```

```python
x_train.shape
```

```python
single_image = x_train[0]
```

# 3. Plot single image to view

```python
plt.imshow(single_image)
```

```python
from tensorflow.keras.utils import to_categorical
```

# 4. Convert target value to one-hot encoding by Tensorflow functionality

```python
y_example = to_categorical(y_train)
```

```python
y_example[0]
```

```python
y_cat_test = to_categorical(y_test, num_classes=10)
```

```python
y_cat_train = to_categorical(y_train, 10)
```

# 5. Scale down data to be from 0 to 1

```python
x_train = x_train / 255
```

```python
x_test = x_test / 255
```

```python
x_train = x_train.reshape(60000, 28, 28, 1)
```

```python
x_test = x_test.reshape(10000, 28, 28, 1)
```

# 6. Import Sequential model from Tensorflow and create simple neuran network

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense, Conv2D, MaxPool2D, Flatten
```

```python
model = Sequential()

model.add(Conv2D(filters=32, kernel_size=(4,4), input_shape=(28, 28, 1), activation='relu'))
model.add(MaxPool2D(pool_size=(2,2)))

model.add(Flatten())

model.add(Dense(128, activation='relu'))

# Output layer Softmax - Multi class
model.add(Dense(10, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam',
             metrics = ['accuracy']
             )
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor='val_loss', patience=1)
```

```python
model.fit(x_train, y_cat_train, epochs=10, validation_data=(x_test, y_cat_test), callbacks=[early_stop])
```

# 7. Model summary

```python
model.summary()
```

```python
metrics = pd.DataFrame(model.history.history)
```

# 8. Plot train and validation loss

```python
metrics[['loss', 'val_loss']].plot()
```

# 9. Plot train and validation accuracy

```python
metrics[['accuracy', 'val_accuracy']].plot()
```

# 10. Evaluate model

```python
model.evaluate(x_test, y_cat_test, verbose=1)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
predict_x = model.predict(x_test) 
predict = np.argmax(predict_x,axis=1)
```

# 11. Classification report

```python
print(classification_report(y_test, predict))
```

```python
confusion_matrix(y_test, predict)
```
