# MODEL CREATION
   I used Convolutiona Neural Network(CNN) for making the classifier.
![](https://miro.medium.com/max/2216/1*0NwaOkzvom6YpMZoIgWTiQ.png)   

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(16,(3,3),activation='relu',input_shape=(28,28,1)),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(32,(3,3),activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(64,(3,3),activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(512,activation='relu'),
    tf.keras.layers.Dense(10,activation='softmax')
])
```

# Dataframe Creation

```python
training_data = pd.read_csv('../input/digit-recognizer/train.csv')
test_data = pd.read_csv('../input/digit-recognizer/test.csv').to_numpy()
training_label = training_data['label'].to_numpy()
training_data = training_data.drop('label',axis=1).to_numpy()
sample_data = pd.read_csv('../input/digit-recognizer/sample_submission.csv')
```

# Model Compilation

```python
model.compile(optimizer='rmsprop',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

# MODEL TRAINING
The final Accuracy obtained is 97%

```python
training_data = training_data.reshape(-1,28, 28, 1)
model.fit(training_data,training_label,epochs=15,validation_data=(training_data[800:1000],training_label[800:1000]))
```

```python
print('training_labels : ',training_label[:5])
```

# TESTING DATA CLASSIFICATION

```python
test_data = test_data.reshape(-1,28, 28, 1)
predictions = model.predict_classes(test_data)
```
