```python
import matplotlib.pyplot as plt
from tensorflow import keras
```

```python
X_train = pd.read_csv('../input/digit-recognizer/train.csv')
X_test = pd.read_csv('../input/digit-recognizer/test.csv')
print(X_train.head())
```

```python
y_train = X_train.pop('label')
print(type(y_train))
```

```python
X_train = (X_train/255).to_numpy()
X_test = (X_test/255).to_numpy()
y_train = y_train.to_numpy()
```

```python
print(type(X_train))
print(X_test.shape)
```

```python
X_train = X_train.reshape(42000,28,28)
```

```python
X_test = X_test.reshape(28000,28,28)
```

```python
print(X_train.shape)
```

```python
print(X_train[0].shape)
X_trainExample = X_train[0].reshape(28,28)
print(X_trainExample.shape)
```

```python
plt.figure()
plt.imshow(X_trainExample,cmap=plt.cm.binary)
plt.show()
```

**Normalizing**

```python
X_train = keras.utils.normalize(X_train, axis=1)
X_test = keras.utils.normalize(X_test, axis=1)
print(X_train)
```

```python
print(y_train)
```

```python
import tensorflow as tf
```

Extend dims of numpy arrays, so they can be used in model

```python
X_train = np.expand_dims(X_train, -1)
X_test = np.expand_dims(X_test, -1)
```

```python
input_shape = (28,28,1)
```

```python
model = keras.models.Sequential()
model.add(keras.layers.Conv2D(32, kernel_size=(3,3), input_shape=input_shape, activation=tf.nn.relu))
model.add(keras.layers.MaxPooling2D(pool_size=(2,2)))
model.add(keras.layers.Conv2D(64, kernel_size=(3,3), activation=tf.nn.relu))
model.add(keras.layers.MaxPooling2D(pool_size=(2,2)))
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(10, activation=tf.nn.softmax))
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
early_stopping = keras.callbacks.EarlyStopping()
model.fit(X_train, y_train, epochs=5, callbacks=[early_stopping])
pred = model.predict(X_test)
```

```python
print(np.argmax(pred[4]))
X_test0 = X_test[4].reshape(28,28)
plt.imshow(X_test0)
plt.show()
print(pred.shape)
```

```python
print(np.argmax(pred, axis=1))
```

```python
subm = pd.DataFrame({'ImageId': np.arange(1, 28001), 'Label':np.argmax(pred, axis=1)})
subm.to_csv('submission.csv',index=False)
print(subm)
```
