```python
import xgboost as xgb #Lets use Graient Boosting, because logreg shows at best 90% accuracy, which is pathetic
```

```python
df_train=pd.read_csv('/kaggle/input/digit-recognizer/train.csv')
df_test=pd.read_csv('/kaggle/input/digit-recognizer/test.csv')
```

```python
X=np.array(df_train.drop(['label'],axis=1),dtype=float)
Y=np.array(df_train['label'],dtype=float)
```

```python
for i in range(3):
    fig=plt.figure()
    plt.imshow(X[i].reshape([28,28]))
    plt.title(str(Y[i]))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X_train,X_test,Y_train,Y_test = train_test_split(X,Y)
```

```python
dtrain = xgb.DMatrix(X_train, label=Y_train)   
dtest = xgb.DMatrix(X_test, label=Y_test)

param = {
    'max_depth': 3,  # max tree depth
    'eta': 0.3,  # learning rate
    'objective': 'multi:softprob',  # multiclass classification
    'eval_metric':'mlogloss', # log loss
    'num_class': 10}  # number of classes
```

```python
N_iter = 120 # boosting iteration num

model = xgb.train(param, dtrain, N_iter)

predictions = model.predict(dtest)

class_preds = np.asarray([np.argmax(line) for line in predictions])

print(accuracy_score(Y_test,class_preds))
```

```python
df_test.index += 1
ImageId = df_test.index
```

```python
test_data = xgb.DMatrix(np.array(df_test))
predictions = model.predict(test_data)
predicted = np.asarray([np.argmax(line) for line in predictions])
```

```python
output = pd.DataFrame({'ImageID': ImageId, 'Label': predicted})
output.to_csv('/kaggle/working/submission.csv', index=False)
```
