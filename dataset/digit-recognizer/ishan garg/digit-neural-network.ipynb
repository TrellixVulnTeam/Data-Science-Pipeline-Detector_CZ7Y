```python
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
font={'family':'sans-serif',
      'size':16}
plt.rc('font',**font)
```

```python
digit_train=pd.read_csv("../input/digit-recognizer/train.csv")
```

```python
digit_train
```

```python
plt.figure(figsize=(17,5))
sns.countplot(digit_train['label'],facecolor=(0,0,0,0),linewidth=3,edgecolor=sns.color_palette("dark",3))
```

```python
plt.figure(figsize=(6,6))
for i in range(9):
    img = np.asarray(digit_train.iloc[i,1:].values.reshape((28,28)));
    plt.subplot(3, 3, i+1)
    plt.imshow(img, cmap='gray')
plt.show()
```

```python
y=digit_train.label
y=pd.get_dummies(y)
X=digit_train.drop('label', axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X,y,train_size=0.9,test_size=0.1,random_state=0)
```

```python
from tensorflow import keras
from tensorflow.keras import layers
```

```python
model = keras.Sequential([
    layers.Dense(units=256, activation='relu', input_shape=[784]),
     layers.Dense(units=10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['categorical_accuracy'])
```

```python
history = model.fit(
    X_train, y_train,
    validation_data=(X_valid, y_valid),
    batch_size=500,
    epochs=10,
    verbose=True
)
```

```python
digit_test=pd.read_csv("../input/digit-recognizer/test.csv")
```

```python
digit_test
```

```python
preds=model.predict(digit_test)
pred_classes = np.argmax(preds,axis=1)
digit_output = pd.DataFrame({'ImageId': digit_test.index+1,
                       'Label': pred_classes})
digit_output
```

```python
digit_output.to_csv('submission.csv', index=False)
```
