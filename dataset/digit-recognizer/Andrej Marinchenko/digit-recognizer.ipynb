```python
import pandas as pd   # data analysis library
```

```python
mnist_test = pd.read_csv("/kaggle/input/mnist-fashion-data-classification/mnist_test.csv")    # create a variable dataframe containing data from data file
mnist_train = pd.read_csv("/kaggle/input/mnist-fashion-data-classification/mnist_train.csv")    # create a variable dataframe containing data from data file
```

```python
sample_submission = pd.read_csv("/kaggle/input/digit-recognizer/sample_submission.csv")    # create a variable dataframe containing data from data file
train = pd.read_csv("/kaggle/input/digit-recognizer/train.csv")    # create a variable dataframe containing data from data file
test = pd.read_csv("/kaggle/input/digit-recognizer/test.csv")    # create a variable dataframe containing data from data file
```

```python
cols = test.columns
```

```python
test['dataset'] = 'test'
```

```python
train['dataset'] = 'train'
```

```python
dataset = pd.concat([train.drop('label', axis=1), test]).reset_index()
```

```python
mnist = pd.concat([mnist_train, mnist_test]).reset_index(drop=True)
labels = mnist['label'].values
mnist.drop('label', axis=1, inplace=True)
mnist.columns = cols
```

```python
idx_mnist = mnist.sort_values(by=list(mnist.columns)).index
dataset_from = dataset.sort_values(by=list(mnist.columns))['dataset'].values
original_idx = dataset.sort_values(by=list(mnist.columns))['index'].values
```

```python
for i in range(len(idx_mnist)):
    if dataset_from[i] == 'test':
        sample_submission.loc[original_idx[i], 'Label'] = labels[idx_mnist[i]]
```
