```python
df=pd.read_csv('../input/digit-recognizer/train.csv')
```

```python
df.head()
```

```python
X_train=df.drop(axis=1,columns=['label'])
```

```python
X_train
```

```python
y_train=df['label']
y_train
```

```python
num_classes = 10 #number of classes, here is 10 (0,1,...,9)
y_train = keras.utils.to_categorical(y_train, num_classes)
```

```python
X_train=X_train/255
```

```python
X_train
```

```python
model=Sequential()
model.add(Dense(64,activation='relu'))
model.add(Dense(256,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=3)
```

```python
test=pd.read_csv('../input/digit-recognizer/test.csv')
```

```python
pred_name = model.predict(test)
```

```python
pred_name
```

```python
Label=np.argmax(pred_name,axis=1)
```

```python
ImageId=np.arange(1,28001)
```

```python
submission=pd.DataFrame(data={"ImageId":ImageId,"Label":Label})
```

```python
submission
```

```python
submission.to_csv('subm.csv', index=False)
```
