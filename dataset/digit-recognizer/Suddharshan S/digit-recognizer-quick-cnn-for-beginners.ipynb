```python
df = pd.read_csv('../input/digit-recognizer/train.csv')
df1 = pd.read_csv('../input/digit-recognizer/test.csv')
y = df['label']
del df['label']
```

```python
x = df.copy()

x = np.array(x).reshape(42000,28,28)
y = np.array(y)

x_train, x_test, y_train, y_test =\
    train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
df1.shape
```

```python
x1 = np.array(df1).reshape(28000,28,28)
```

```python
x.shape
```

```python
num_classes = 10
img_height = 28
img_width = 28
```

```python
model = Sequential([
  layers.Rescaling(1./255, input_shape=(img_height, img_width, 1)),
  layers.Conv2D(16, (5,5), padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(32, (5,5), padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(64, (5,5), padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(128, (5,5), padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(128, (5,5), padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Flatten(),
  layers.Dense(256, activation='relu'),
  layers.Dense(num_classes)
])
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
history = model.fit(x,y,epochs=500,verbose=False)
```

```python
pred = model.predict(x1)
```

```python
pred = pred.argmax(axis=1)
```

```python
sub = pd.DataFrame({'ImageId':range(1,28001),'Label':pred})
sub.to_csv('submission.csv',index=False)
```
