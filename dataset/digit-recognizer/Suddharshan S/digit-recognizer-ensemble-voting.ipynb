```python
x = df.copy()
```

```python
x_train, x_test, y_train, y_test =\
    train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
clf1 = CatBoostClassifier(iterations=500,verbose=False)
clf2 = XGBClassifier(estimators=100)
clf3 = LGBMClassifier(estimators=150)

model = VotingClassifier(estimators=[
 ('lr', clf1), ('rf', clf2), ('gnb', clf3)], voting='soft')
```

```python
model.fit(x,y)
pred = model.predict(df1)
```

```python
pred = pd.Series(pred.flatten())
```

```python
sub = pd.DataFrame({'ImageId':range(1,28001),'Label':pred})
sub.to_csv('submission.csv',index=False)
```
