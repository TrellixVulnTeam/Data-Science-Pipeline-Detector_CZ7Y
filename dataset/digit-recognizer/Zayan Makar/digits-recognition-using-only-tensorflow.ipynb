**Import Library**

```python

import tensorflow as tf
import warnings
warnings.simplefilter('ignore')
import pandas as pd
from sklearn.model_selection import train_test_split
```

**Read Data**

```python

traindata=pd.read_csv("../input/digit-recognizer/train.csv")
y=traindata['label']
X=traindata.drop('label',axis=1)
```

**Split Train Data**

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.21,random_state=44)
```

**Create Neural Network**

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(512, activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(128, activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(10, activation=tf.nn.softmax))
```

**Data Normalization**

```python
x_train = tf.keras.utils.normalize(x_train, axis=1)
x_test = tf.keras.utils.normalize(x_test, axis=1)
```

**Complie**

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=30)
```

```python
val_loss, val_acc = model.evaluate(x_test, y_test)
print(val_loss)
print(val_acc)
```

```python
y_pred = model.predict(x_test)
```

**Predict data from  test **

```python
testdata=pd.read_csv("../input/digit-recognizer/test.csv")
X_test=testdata
X_test = tf.keras.utils.normalize(X_test, axis=1)
y_pred=model.predict(X_test)
```

```python
sample_submission=pd.read_csv("../input/digit-recognizer/sample_submission.csv")
ImageId=sample_submission['ImageId']
import numpy as np
y_pred2=[]
for i in range(len(y_pred)):
    y_pred2.append(np.argmax(y_pred[i]))  

```

```python
sample_submission = pd.DataFrame({'ImageId': ImageId,'Label': y_pred2})
print(sample_submission)
sample_submission.to_csv('sample_submission.csv',index=False)
```
