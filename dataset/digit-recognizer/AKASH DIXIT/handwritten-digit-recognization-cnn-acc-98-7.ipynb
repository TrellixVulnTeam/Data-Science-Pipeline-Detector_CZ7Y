```python
import pandas as pd
from matplotlib import pyplot
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from numpy import asarray, unique, argmax
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPool2D, Flatten, Dropout
from keras.optimizers import RMSprop
```

```python
#Reading the train dataset
data = pd.read_csv('/kaggle/input/digit-recognizer/train.csv')
data.head()
```

```python
#checking the shape of provided train dataset
data.shape
```

```python
#checking unique numbers in train label column
unique = data['label'].unique()
print("Unique Numbers :",unique)

#countine the unique number of digits for classification
n_classes = len(unique)
print("Number of classes :",n_classes)
```

```python
#Filtering the dataset between X and Y
x = data.drop(labels = ["label"], axis=1)
print("X= ",x)
```

```python
#Filtering the target variable from the train data
y = data['label']
print("Target Variable: ",y)
```

```python
#Splitting the dataset in train and test
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.20, random_state=42)
```

```python
#Priting the Splitted dataset
print("X_train: ", x_train)
print("X_test: ", y_test)
print("Y_train: ", y_train)
print("Y_test: ", y_test)
```

```python
#Reshaping the Dataset
x_train = x_train.values.reshape(-1,28,28,1)
x_test = x_test.values.reshape(-1,28,28,1)
```

```python
#determine the shape of input images
in_shape = x_train.shape[1:]
print(in_shape)
```

```python
plt.imshow(x_train[0].reshape([28,28]))
```

```python
# normalize pixel values
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0
```

```python
model = Sequential()

model.add(Conv2D(filters = 32, kernel_size = (5,5),padding = 'Same', 
                 activation ='relu', input_shape = in_shape))
model.add(Conv2D(filters = 32, kernel_size = (5,5),padding = 'Same', 
                 activation ='relu'))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Dropout(0.25))


model.add(Conv2D(filters = 64, kernel_size = (3,3),padding = 'Same', 
                 activation ='relu'))
model.add(Conv2D(filters = 64, kernel_size = (3,3),padding = 'Same', 
                 activation ='relu'))
model.add(MaxPool2D(pool_size=(2,2), strides=(2,2)))
model.add(Dropout(0.25))


model.add(Flatten())
model.add(Dense(256, activation = "relu"))
model.add(Dropout(0.5))
model.add(Dense(10, activation = "softmax"))
```

```python
optimizer = RMSprop(lr=0.001, rho=0.9, epsilon=1e-08, decay=0.0)
```

```python
# define loss and optimizer
model.compile(optimizer= 'adam', loss = 'sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# fit the model
model.fit(x_train, y_train, validation_split=0.2, epochs=100, batch_size=128, verbose= 1)
```

```python
#Evaluate the model
loss, accuracy = model.evaluate(x_test, y_test, verbose=1)
print('Accuracy: %.3f' %accuracy)
print('Loss: ',loss)
```

```python
#make a pridiction
image = x_test[2]
ypred = model.predict(asarray([image]))
print('Prediction: Class =%d' %argmax(ypred))
```
