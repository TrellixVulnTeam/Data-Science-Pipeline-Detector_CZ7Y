Anyway, let's load datasets.

```python
import pandas as pd
train_data=pd.read_csv("/kaggle/input/digit-recognizer/train.csv")
test_data=pd.read_csv("/kaggle/input/digit-recognizer/test.csv")
```

<pre>
Let's show train_data.
</pre>

```python
train_data.head()
```

<pre>
That means the first column is "label", columns from the 2nd to the last is "28 x 28 = 784 pixcels".
</pre>

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(30,30))
for i in range(10):
    plt.subplot(20, 20, i+1)
    plt.title("No." + str(i))
    plt.imshow(train_data.iloc[:,1:].iloc[i].values.reshape(28,28),cmap='Greys')
```

<pre>
Okay, let's split train_data into x_train, y_train, x_valid, y_valid.
</pre>

```python
from sklearn.model_selection import train_test_split
import numpy as np
x=np.array(train_data.iloc[:,1:785])
y=np.array(train_data["label"])
x_train, x_valid, y_train, y_valid = train_test_split(x,y,test_size=0.3, random_state=42)
```

<pre>
Let's start sklearn.neural_network MLPClassifier.
</pre>

```python
from sklearn.neural_network import MLPClassifier
deep_learning = MLPClassifier(random_state=123,verbose=True)
deep_learning.fit(x_train, y_train)
```

<pre>
Let's show the model's loss curve.
</pre>

```python
plt.plot(deep_learning.loss_curve_)
```

<pre>
Let's test the model with validation data.
</pre>

```python
valid_test=deep_learning.predict(x_valid)
```

```python
valid_test_result=pd.concat([pd.DataFrame(valid_test),pd.DataFrame(y_valid)], axis=1)
valid_test_result.columns=["prediction","correct label"]
valid_test_result
```

<pre>
Okay, well done!

Now is the time for prediction.
</pre>

<pre>
Try the following sample.
</pre>


```python
x_test=np.array(test_data)
plt.imshow(x_test[[0]].reshape(28,28),cmap='Greys')
```

<pre>
It's 2.
</pre>

```python
y_pred_sample= deep_learning.predict(x_test[[0]])
print("The answer is",int(y_pred_sample),".")
```
