# MNIST digit recognition with Random Forests
In this kernel I will try to predict digits from the MNIST database using a Random Forest Classifier. 

```python
import pandas as pd
import numpy as np
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../input/train.csv")
```

```python
data.head()
```

```python
L = np.sqrt(784)
L
```

### Plot a few images

```python
def plotNum(ind):
    plt.imshow(np.reshape(np.array(data.iloc[ind,1:]), (28, 28)), cmap="gray")
```

```python
plt.figure()
for ii in range(1,17):
    plt.subplot(4,4,ii)
    plotNum(ii)
```

### Process data

```python
X = data.iloc[:, 1:]
y = data['label']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_jobs=-1, n_estimators=10)
```

```python
rfc.fit(X_train, y_train)
```

```python
rfc.score(X_test, y_test)
```
