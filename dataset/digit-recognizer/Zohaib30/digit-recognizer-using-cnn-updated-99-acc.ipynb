# The network architecture

```python
network = models.Sequential()
# network.add(layers.Dense(50, activation='relu', kernel_initializer='normal', input_shape=(28 * 28,)))
# network.add(layers.BatchNormalization())
# network.add(layers.Dense(50, activation='relu', kernel_initializer='normal'))
# network.add(layers.BatchNormalization())
# network.add(layers.Dense(50, activation='relu', kernel_initializer='normal'))
# network.add(layers.BatchNormalization())
# network.add(layers.Dense(10, activation='softmax',  kernel_initializer='normal', ))
network.add(layers.Conv2D(32, (3,3), activation='relu', input_shape=(28, 28, 1)))
network.add(layers.MaxPooling2D((2, 2)))
network.add(layers.Conv2D(64, (3,3), activation='relu'))
network.add(layers.MaxPooling2D(2,2))
network.add(layers.Conv2D(64, (3,3), activation='relu'))
network.add(layers.Flatten())
network.add(layers.Dense(64, activation='relu'))
network.add(layers.Dense(10, activation='softmax'))
```

```python
network.summary()
```

```python
train_image = train_image.reshape((60000, 28 , 28, 1))
train_image = train_image.astype('float32') / 255

test_images = test_images.reshape((10000, 28 , 28, 1))
test_images = test_images.astype('float32') / 255
```

```python
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
```

```python
network.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history = network.fit(train_image, train_labels, epochs=60)
```

```python
def plot_loss(loss, acc):
    plt.figure()
    plt.plot(loss)
    plt.plot(acc)
    plt.title('Model loss')
    plt.xlabel('Epoch')
    plt.legend(['Traning Loss', 'Traning Accuracy'], loc='upper right')
    plt.show()
```

```python
plot_loss(history.history['loss'], history.history['accuracy'])
```

```python
test_loss, test_acc = network.evaluate(test_images, test_labels)
print('test_acc: ', test_acc)
```
