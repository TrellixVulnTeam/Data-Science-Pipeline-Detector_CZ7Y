```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import  numpy as np
from keras.utils.np_utils import to_categorical
```

# Load data

```python
train=pd.read_csv('../input/digit-recognizer/train.csv')
test=pd.read_csv('../input/digit-recognizer/test.csv')
```

```python
train.head(3)
```

```python
train.shape
```

```python
test.shape
```

```python
sns.countplot(train['label'])
```

```python
x_train=train.drop(['label'],1)
y_train=train['label']
```

```python
x_train=np.array(x_train)
test=np.array(test)
```

```python
x_train.shape
```

```python
test.shape
```

```python
x_train=x_train.reshape(42000,28,28)
test=test.reshape(28000,28,28)
```

```python
x_train=x_train/255
test=test/255
```

```python
def show_img(i):
    plt.imshow(x_train[i],cmap='gray')
    plt.xlabel(y_train[i])
show_img(0)
```

```python
target=x_train.reshape(-1,28,28,1)
test=test.reshape(-1,28,28,1)
y_train=np.array(y_train)
label=to_categorical(y_train)
label.shape
```

# Model creation

```python
from keras.models import Sequential
from keras.layers import Conv2D,MaxPooling2D,Flatten,Dense,Dropout
```

```python
model=Sequential([
    Conv2D(32,(5,5) , activation='relu' , input_shape=(28,28,1)),
    MaxPooling2D(pool_size=(2,2)),
    
    Conv2D(64,(5,5), activation ='relu'),
    MaxPooling2D(pool_size=(2,2)),
    Dropout(0.25),
    
    Conv2D(64,(3,3), activation ='relu'),
    MaxPooling2D(pool_size=(2,2)),
    Dropout(0.25),
    
    Flatten(),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(target,label,epochs=40,batch_size=64)
```

```python
Y_pred = model.predict(test)
Y_pred_classes = np.argmax(Y_pred,axis = 1) 
```

```python
submission_data = pd.read_csv('../input/digit-recognizer/sample_submission.csv')
```

```python
submission_data['Label']=Y_pred_classes
```

```python
submission_data.to_csv('submit.csv' ,index=False)
```
