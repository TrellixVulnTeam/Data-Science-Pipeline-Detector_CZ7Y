```python
digit=pd.read_csv('/kaggle/input/digit-recognizer/train.csv')
digit               
```

```python
import matplotlib.pyplot as plt 
plt.imshow(digit.iloc[2002,1:].to_numpy().reshape(28,28))
```

```python
digit.iloc[50,1:].to_numpy().reshape(1,784)
```

```python
x=digit.drop(columns=['label'])
y=digit.label
x.shape
```

```python
x=x/255
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Flatten
model=Sequential()
model.add(Dense(4,activation='sigmoid',input_shape=(784,)))
model.add(Dense(5,activation='relu'))
model.add(Dense(3,activation='softmax'))
model.summary()
```

```python
model.compile(loss='hinge',optimizer='adam')
```

```python
model.fit(x_train,y_train,epochs=2)
```

## Using mnist dataset

```python
from tensorflow import keras 
(x_train,y_train),(x_test,y_test)=keras.datasets.mnist.load_data()          #inbuilt data 
```

```python
x_train.shape
```

```python
import matplotlib.pyplot as plt
plt.imshow(x_train[1])
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Flatten
model=Sequential()
model.add(Flatten(input_shape=(28,28)))
model.add(Dense(4,activation='sigmoid'))
model.add(Dense(5,activation='relu'))
model.add(Dense(10,activation='softmax'))
model.summary()
```

```python
import tensorflow as tf
model.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=30)
```

```python
pred=model.predict(x_train)
pred
```
