## function : x**2  
<a id = 'section_1'></a>

## 1) Gradient Descent  
<a id = 'section_1_1'></a>

```python
def function(x):
    return x**2
```

We know the formula for the derivative (f(x + delta(x)) - f(x))/ delta(x).  
You can get good idea on this page [What exactly gradient is?](https://www.mathsisfun.com/calculus/derivatives-introduction.html)

```python
# We will calculate by passing a value of x and delta x, I have choosen as delta x = 0.0001
def derivative(x):
    delta_x = 0.0001
    return (function(x + delta_x) - function(x))/ delta_x 
```

We will try to get points of a curve x^2

```python
x = np.arange(-100,100,0.1)
y = function(x)
```

```python
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('X^2 curve')
plt.show()
```

```python
x_value = -100
lr = 0.01
new_points = [] # In this we will strore all update of initial point
for i in range(200):
    x_value  = x_value - lr * derivative(x_value)
    new_points.append(x_value)
```

```python
plt.plot(x, y)
plt.scatter(new_points, function(np.array(new_points)),s = 10, c='r')
plt.xlim(-100, 100)
plt.show()
```

### Animation  


Let's visualize it so that, We can understand how we are decreasing error with iteration

```python
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
import numpy as np
import matplotlib.pyplot as plt
import math
def Create_animation(new_points):
    fig = plt.figure()

    plt.plot(x, y, lw = 2)
    # plt.scatter(new_points, f(np.array(new_points)),s = 10, c='r')

    balle, = plt.plot([0], [0], 'ko', ms = 10, mfc = 'r')

    def animate(i):

        balle.set_data(i, function(i))
        return balle,
    
    animation = FuncAnimation(fig, animate, frames = np.array(new_points), interval = 100,blit =True, repeat=True)
    return animation
animation = Create_animation(new_points)    
HTML(animation.to_html5_video())
```

As we see from the animation above, It our ball is taking too much time to reach global minima, It stops when it reached  
to the minima.

## function:- x*cos(1.2*x)  
<a id = 'section_2'></a>

## 1) Gradiant descent  
<a id = 'section_2_1'></a>

```python
def function(x):
    return x*tf.math.cos(1.2*x)

# We will calculate by passing a value of x and delta x, I have choosen as delta x = 0.0001
def derivative(x):
    delta_x = 0.0001
    return (function(x + delta_x) - function(x))/ delta_x  # This is formula we use to calculate derivative of a point

x = np.arange(-4,4,0.1)
y = function(x)
```

```python
plt.plot(x,y)
plt.title('Curve with mathematical equation = x * cos(1.2 * x)')
plt.show()
```

Let's try to play around with this curve where we have 2 local minima, Our goal is to reach the Global minima.  

Conclusion from the experiments that i have done  
i) There are two minima and our point just get stuck at a 1st minima. I have tried with higher learning rate but still  
  Unable to leave the 1st minima.  
ii) Also tried with increasing iteration, But didn't help.  

Becauuuuuuseeeee, We reached at zero gradient.  

```python
x_value = -2.5
new_points = []
for i in range(200):
    x_value  = x_value - 0.01 * derivative(x_value)
    new_points.append(x_value)
# new_points = [i.numpy() for i in new_points]
```

```python
plt.plot(x, y)
plt.scatter(new_points, function(np.array(new_points)),s = 15, c='r')
plt.xlim(-4, 4)
plt.title('Showing the updation we achieved during iterations')
plt.show()
```

### Animation

## 2)  Gradient Descent + Momentum  
<a id = 'section_2_2'></a>

### 2.1) Hyperparameters :   
ro = 0.9  
learning_rate = 0.01  

As we can see due to the velocity, Our ball escaped the first local minima, And went little bit up

```python
x_value = -2.5
vx = 0    # Initializing velocity with zero
rho = 0.9 # As per documentation, Value of rho can vary between 0.9 - 0.99
new_points = []
for i in range(300):
    d = derivative(x_value)
    vx = rho * vx + d
    x_value  = x_value - 0.01 * vx
    new_points.append(x_value)
# new_points = [i.numpy() for i in new_points]
```

```python
plt.plot(x, y)
plt.scatter(new_points, function(np.array(new_points)),s = 20, c='r')
plt.xlim(-4, 10)
plt.title('Showing the updation we achieved during iterations')
plt.show()
```

### Animation

```python
animation = Create_animation(new_points)    
HTML(animation.to_html5_video())
```

### 2.2) HyperParameters :  
Yehhhhhh, Finally after some hyper parameter tuning, able to show that this optimizer is using velocity.  
ro = 0.95,  learning_rate = 0.02   

* Comments:-  
As you can see that, Velocity is increasing with time as we move on slope, Also we can see oscillation of ball due to  
momentum of the ball.

```python
x_value = -2.5
vx = 0
rho = 0.95
new_points = []
velocity = []
for i in range(200):
    d = derivative(x_value)
    velocity.append(vx)
    vx = rho * vx + d
    x_value  = x_value - 0.02 * vx
    new_points.append(x_value)
# new_points = [i.numpy() for i in new_points]
```

```python
plt.plot(x, y)
plt.scatter(new_points, function(np.array(new_points)),s = 10, c='r')
plt.xlim(-4, 4)
plt.title('Showing the updation we achieved during iterations')
plt.show()
```

This will be really interesting to see how we velocity is chsnging over time,  
Let's visualize it for better understanding  

```python
plt.plot(x, y)
plt.plot(np.linspace(-2.5, 4,200), velocity, c='r')
plt.title('Velocity over time')
plt.xlabel('Time- Axis')
plt.ylabel('Velocity')
plt.show()
```

## 3) Nestorom Momentum   
<a id = 'section_2_3'></a>

```python
x_value = -2.5
v = 0
rho = 0.95
lr = 0.03
new_points = []
for i in range(200):
    dx = derivative(x_value)
    old_v = v
    v = rho*v - lr * dx
    x_value  = x_value - (rho * old_v) + (1+rho)*v
    new_points.append(x_value)
# new_points = [i.numpy() for i in new_points]
```

```python
plt.plot(x, y)
plt.scatter(new_points, function(np.array(new_points)),s = 10, c='r')
plt.title('Showing the updation we achieved during iterations')
plt.xlim(-4, 10)
plt.show()
```

### Animaion  
<a id = 'animationNM'></a>

```python
animation = Create_animation(new_points)    
HTML(animation.to_html5_video())
```

## 4) AdaGrad  
<a id = 'section_2_4'></a>

```python
grad_square = 0
x_value = -2.5
lr= 0.09
new_points = []
for i in range(200):
    dx = derivative(x_value)
    grad_square = grad_square + dx**2
    x_value = x_value - lr * dx/(np.sqrt(grad_square) + 0.0000001)
    new_points.append(x_value)
# new_points = [i.numpy() for i in new_points]
```

```python
plt.plot(x, y)
plt.scatter(new_points, function(np.array(new_points)),s = 10, c='r')
plt.title('Showing the updation we achieved during iterations')
plt.xlim(-4, 4)
plt.show()
```

### Animaion  
<a id = 'animationA'></a>

```python
animation = Create_animation(new_points)    
HTML(animation.to_html5_video())
```

## 5) RMSProp   
<a id = 'section_2_5'></a>

```python
x_value = -2.5
grad_square = 0
decay_rate = 0.95
lr= 0.01
new_points = []
for i in range(200):
    dx = derivative(x_value)
    grad_square = decay_rate * grad_square + (1-decay_rate)*dx*dx
    x_value = x_value - (lr * dx)/(np.sqrt(grad_square)+ 0.00000001)
    new_points.append(x_value)
# new_points = [i.numpy() for i in new_points]
```

```python
plt.plot(x, y)
plt.scatter(new_points, function(np.array(new_points)),s = 10, c='r')
plt.xlim(-4, 4)
plt.show()
```

## 6) Adam  
<a id = 'section_2_6'></a>

```python
first_moment = 0
second_moment = 0
beta_1 = 0.98
beta_2 = 0.999
lr = 0.01

x_value = -2.5
new_points = []
for i in range(1,900):
    dx = derivative(x_value)
    first_moment = beta_1 * first_moment +(1-beta_1)*dx
    second_moment = beta_2 * second_moment + (1-beta_2)*dx*dx
    x_value = x_value - (lr*first_moment)/(np.sqrt(second_moment)+ 0.0000001)
    new_points.append(x_value)
# new_points = [i.numpy() for i in new_points]
```

```python
plt.plot(x, y)
plt.scatter(new_points, function(np.array(new_points)),s = 10, c='r')
plt.xlim(-4, 4)
plt.show()
```
