# Importing libraries

```python
import tensorflow as tf
from keras.layers import Dense,Activation,Dropout
from keras.models import Sequential
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

# Reading the Data

```python
data = pd.read_csv('../input/digit-recognizer/train.csv')
```

```python
data.head()
```

```python
x_train = data.drop(columns=['label'])      # Reading the images by not including the labels
y_train = data['label']                     # Reading all the labels
```

**Splitting the data into train and validation steps** 
> In general it is not mandatory to create a validation set. But inorder to assess our model's performance we do this step. In practice the ratio of splits is 80:20 if data is in scale of thousands but we do 90:10 if data length is in order of million

```python
len(data)
```

```python
x_train,x_val,y_train,y_val=train_test_split(x_train,y_train,test_size=0.10)
```

# Creating the model
> This is the major step which decides how well should the model perform. Excessive additon of layers would cause more time to train and more memory leading to less speed of our algorithm. So choosing the model is very important. After a few trials I found this to work good. You can try your own.

```python
model=Sequential()
model.add(Dense(512,activation='relu',input_shape=(784,)))
model.add(Dense(256,activation='relu'))
model.add(Dense(128,activation='relu'))
model.add(Dense(64,activation='relu'))
model.add(Dense(32,activation='relu'))
'''model.add(Dense(25,activation='relu'))
model.add(Dense(25,activation='relu'))
model.add(Dense(10,activation='relu'))'''
model.add(Dense(10,activation='softmax'))
model.summary()
```

**After many trails I found this to work. I have used categorical cross entropy as loss and ADAM optimizer. My evaluation metrics being Accuracy**

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history = model.fit(x_train,y_train,epochs=100,batch_size=256)
```

**Plotting the loss vs Accuracy graph**

```python
plt.plot(history.history['accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train'], loc='upper left')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.title('model Loss')
plt.ylabel('Loss')
plt.xlabel('epoch')
plt.legend(['train'], loc='upper left')
plt.show()
```

**Evaluating our Model's Performance over the validation data**

```python
model.evaluate(x_val,y_val)
```
