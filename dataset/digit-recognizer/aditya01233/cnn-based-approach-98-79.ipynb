```python
df_train=pd.read_csv("../input/digit-recognizer/train.csv")
```

```python
df_train.head()
```

```python
x_train=df_train.drop(columns=['label'])
y_train=df_train['label']
```

```python
x_train=x_train.values
y_train=y_train.values
```

```python
print(df_train.shape,x_train.shape,y_train.shape)
```

```python
x_train=np.reshape(x_train,(42000,28,28,1))
```

```python
x_train.shape,y_train.shape
```

```python
mean_px = x_train.mean().astype(np.float32)
std_px = x_train.std().astype(np.float32)
```

```python
x_train=(x_train-mean_px)/std_px
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x_train, y_train, test_size=0.13, random_state=42)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,Dropout
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Flatten
from tensorflow.keras.optimizers import SGD
```

```python
import tensorflow as tf
```

```python
def define_model():
    model = Sequential()
    model.add(Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_uniform', input_shape=(28, 28, 1)))
    model.add(Conv2D(64, (3, 3), activation='relu', kernel_initializer='he_uniform'))
    model.add(MaxPooling2D((2, 2)))
    model.add(Flatten())
    model.add(Dense(100, activation='relu', kernel_initializer='he_uniform'))
    model.add(Dense(10, activation='softmax'))
    opt = SGD(learning_rate=0.01, momentum=0.9)
    model.compile(optimizer=opt, loss=tf.keras.losses.SparseCategoricalCrossentropy(), metrics=['accuracy'])
    return model
```

```python
model=define_model()
history = model.fit(X_train, Y_train, epochs=20, batch_size=32, validation_data=(X_test, Y_test))
```

```python
df_test=pd.read_csv("../input/digit-recognizer/test.csv")
```

```python
mean_px.shape,std_px.shape
```

```python
x_test=df_test.values
```

```python
x_test.shape
```

```python
x_test=np.reshape(x_test,(28000,28,28,1))
```

```python
x_test=(x_test-mean_px)/std_px
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred=y_pred.argmax(axis=1)
```
