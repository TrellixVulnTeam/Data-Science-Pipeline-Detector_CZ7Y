```python
df = pd.read_csv('../input/digit-recognizer/train.csv')
```

```python
y_train = df['label']
x_train = df.drop("label",axis=1)
```

```python
x_train=np.array(x_train)
x_train= x_train.reshape(len(x_train),28,28,1)
```

```python
for k in range (0,len(x_train)):
    for i in range(0,28):
        for j in range(0,28):
            if x_train[k][i][j]>25:
                x_train[k][i][j]=255
            else:
                x_train[k][i][j]=0
```

```python
from keras.layers import Dense, Activation, Dropout, Conv2D, MaxPooling2D, Flatten
from keras.models import Sequential
model = Sequential()
model.add(Conv2D(32, kernel_size=(5, 5), strides=(1, 1),activation='tanh',input_shape=(28, 28, 1)))
model.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))
model.add(Conv2D(64, (5, 5), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))
model.add(Flatten())
model.add(Dense(1000, activation='relu'))
model.add(Dense(10, activation='softmax'))
```

```python
import keras
model.compile(loss=keras.losses.sparse_categorical_crossentropy,
              optimizer=keras.optimizers.SGD(lr=0.01),
              metrics=['accuracy'])
```

```python
model.fit(np.array(x_train), y_train, batch_size=30, epochs=30, verbose=1)
```

```python
df2 = pd.read_csv('../input/digit-recognizer/test.csv')
```

```python
x_test = np.array(df2)
x_test= x_test.reshape(len(x_test),28,28,1)
```

```python
for k in range (0,len(x_test)):
    for i in range(0,28):
        for j in range(0,28):
            if x_test[k][i][j]>25:
                x_test[k][i][j]=255
            else:
                x_test[k][i][j]=0
```

```python
pre=[]
pred = model.predict(np.array(x_test))
for i in range(0,len(x_test)):
    p=pred[i][0]
    tmp=0
    for j in range(0,10):
        if pred[i][j]>p:
            p=pred[i][j]
            tmp=j
    pre.append(tmp)
```

```python
plt.figure()
f, axarr = plt.subplots(4,6) 

for i in range(0,4):
    for j in range(0,6):
        axarr[i][j].imshow((x_test[(6*i)+j]).reshape(28,28))
```

```python
lst=[]
for i in range(0, len(x_test)):
    lst.append((i+1, pre[i]))
```

```python
import pandas
df = pandas.DataFrame(data=lst)
df.to_csv("./file.csv", sep=',',index=True)
```
