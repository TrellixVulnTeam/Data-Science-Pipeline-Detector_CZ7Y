```python
train = pd.read_csv('/kaggle/input/walmart-recruiting-store-sales-forecasting/train.csv.zip')
# 보통 train 데이터셋에는 컬럼이 하나더 많다. weekly_sales가 정답 컬럼이기 때문에.
train
```

```python
# zip 압축되어 있는 파일도 내부적으로 압축 해제하여 읽어들인다.
# 내부에 파일이 여러개인 경우에는 오류 발생
test = pd.read_csv('/kaggle/input/walmart-recruiting-store-sales-forecasting/test.csv.zip')
test
```

```python
# 가장 중요한 전처리 2가지 : 숫자가 아닌 것들을 숫자로 변경(혹은 제거). 차원의 일치

# 먼저 데이터타입 확인해보자
train.dtypes
# 오브젝트는 숫자가 아닌 컬럼이므로 처리가 필요!
```

```python
# 차원이 같아야 한다. 컬럼이 다르면 컬럼제거
# 새로운 변수에 저장해서 원본 보존
# 다중컬럼 제거시에는 리스트 형태로.
# 여기서는 train의 정답 컬럼은 학습을 위해 제거해야 하고, Date는 숫자 형태가 아니므로 제거한다.(변경할 수 있다면 변경도 시도해보아야 함)
train2 = train.drop(columns=['Weekly_Sales', 'Date'])
train2
```

```python
# train과 동일하게 처리되어야 함
test2 = test.drop(columns='Date')
test2
```

```python
# 모델의 선정이 중요하다. 어떤 모델을 선택할 것인가.
# 성능이 잘 나오는 모델로 처음에는 시작하는 것이 좋다.
# 랜덤포레스트가 그 중 하나

from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(train2, train['Weekly_Sales']) # 데이터 패턴과 정답을 전달하여 학습시킴
```

```python
# 이제는 학습된 것들로 예측을 하여 결과값을 리스트로 받자
result = rf.predict(test2)
result
```

```python
test['Weekly_Sales'] = result
test.head(50)
```

```python
sub['Weekly_Sales'] = result
sub
```
