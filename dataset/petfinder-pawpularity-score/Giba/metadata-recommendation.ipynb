```python
train = pd.read_csv('../input/petfinder-pawpularity-score/train.csv')
train
```

```python
train.columns
```

```python
metadata = ['Subject Focus', 'Eyes', 'Face', 'Near', 'Action', 'Accessory', 'Group', 'Collage', 'Human', 'Occlusion', 'Info', 'Blur']

MEAN = train.Pawpularity.mean()
COUNT = 100
dt = train.groupby(metadata)['Pawpularity'].agg(['mean', 'count']).reset_index()

# Smooth the mean by taking into acocunt the frequency
dt['likelihood'] = ((dt['mean']*dt['count']) + (MEAN*COUNT)) / (dt['count']+COUNT) 
del dt['mean'], dt['count']
dt.head(10)
```

```python
# All meta zeroed have Pawpularity = 33.045817
dt.head(1)
```

```python
# Sort by Pawpularity
dt = dt.sort_values('likelihood', ascending=False).reset_index(drop= True)
dt.head(20)
```
