# Petfinder Data Arrange

Petfinder: https://www.petfinder.my/

Pet images are $3 \times 3$ gird patterns, but one image is given in the dataset. So I make grid image patterns and target lists.

```python
import os
import gc
import warnings

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import cv2

warnings.filterwarnings("ignore")
sns.set()

%matplotlib inline
```

## Data Load

```python
DATAPATH = "../input/petfinder-pawpularity-score"
TRAIN_IMGPATH = "../input/petfinder-pawpularity-score/train"
TEST_IMGPATH = "../input/petfinder-pawpularity-score/test"
train_df = pd.read_csv(os.path.join(DATAPATH, "train.csv"))
test_df = pd.read_csv(os.path.join(DATAPATH, "test.csv"))
subission = pd.read_csv(os.path.join(DATAPATH, "sample_submission.csv"))

print(f"train data shape: {train_df.shape}")
print(f"test data shape: {test_df.shape}")
```

## Helper Function  

```python
# shape[9] images -> shape[3, 3] images
def concat_tile(im_list_2d):
    return cv2.resize(cv2.vconcat([cv2.hconcat(im_list_h) for im_list_h in im_list_2d]), (224, 224))
```

```python
# Get random index to select random image
def random_idx(exc_idx, end_idx):
    idx_list = np.zeros(9)
    while (len(idx_list) != len(set(idx_list))) and (exc_idx not in idx_list):
        idx_list = np.random.randint(0, end_idx, 9)

    return idx_list
```

```python
# Get random index to set target image's position
def get_pos():
    return np.random.randint(0, 9)
```

```python
# Get grid images and a target list
def make_grid(target_idx):
    end_idx = train_df.shape[0]
    pos = get_pos()
    img_idx_list = random_idx(target_idx, end_idx)
    img_idx_list[pos] = target_idx

    img_list = np.array([cv2.resize(cv2.imread(os.path.join(TRAIN_IMGPATH, train_df.loc[i, "Id"]+".jpg")),
                                    (224, 224)) for i in img_idx_list]).reshape(3, 3, 224, 224, 3)
    img = concat_tile(img_list)
    target_list = train_df.loc[img_idx_list, "Pawpularity"].values

    return img, target_list, img_idx_list
```

```python
img, target, idx = make_grid(4)
```
