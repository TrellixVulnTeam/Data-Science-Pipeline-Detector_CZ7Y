```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Path
BASE_PATH  = "../input/petfinder-pawpularity-score/"
TRAIN_PATH = BASE_PATH + "train.csv"
TEST_PATH  = BASE_PATH + "test.csv"

train_df = pd.read_csv(TRAIN_PATH)
```

```python
plt.figure(figsize=(20,3))
plt.title('Pawpularity')
plt.hist(train_df["Pawpularity"], bins=100, color="#01C5C4")
plt.show()
```

```python
target_df = train_df.sort_values(by=['Pawpularity', 'Id'], ascending=[False, True]).reset_index(drop=True)
target_df
```

```python
plt.figure(figsize=(32, 320))
for index, column_name in target_df[:100].iterrows():
    col_id          = column_name['Id']
    col_Pawpularity = column_name['Pawpularity']
    
    plt.subplot(25, 4, index+1)
    path = f'../input/petfinder-pawpularity-score/train/{col_id}.jpg'
    im = plt.imread(path)
    plt.imshow(im)
    plt.title(f'Pawpularity: {int(col_Pawpularity)}')
    plt.axis('off')
```
