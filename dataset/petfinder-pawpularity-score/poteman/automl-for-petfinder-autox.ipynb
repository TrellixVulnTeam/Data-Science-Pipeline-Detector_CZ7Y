In this notebook, we demonstrate how AutoX handle task with image data.

Since petfinder is a kernel only competition, we split the original train data into train and test.
For details of data splition, please check [petfinder-data-process](https://www.kaggle.com/poteman/petfinder-data-process)

```python
from autox import AutoX

path = f'../input/petfinder-train-test-split'
image_info = {
    'image_path': '../input/petfinder-pawpularity-score/train',
    'image_col': 'Id',
    'filename_extension': 'jpg'
}

autox = AutoX(target = 'Pawpularity', train_name = 'train.csv', test_name = 'test.csv', 
               id = ['Id'], path = path,  image_info = image_info)
```

```python
sub = autox.get_submit()
```

```python
sub.head()
```

```python
import pandas as pd
solution = pd.read_csv('../input/petfinder-train-test-split/solution.csv')
```

```python
y_true = solution['Pawpularity']
y_pred = sub['Pawpularity']
```

```python
from sklearn.metrics import mean_squared_error

# squared: If True returns MSE value, if False returns RMSE value.
rmse = mean_squared_error(y_true, y_pred, squared = False)
print(rmse)
```
