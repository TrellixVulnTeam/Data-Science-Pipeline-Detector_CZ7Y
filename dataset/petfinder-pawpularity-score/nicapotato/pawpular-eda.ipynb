```python
df = pd.read_csv("../input/petfinder-pawpularity-score/train.csv")

with nicaviz.timer("Reduce Memory"):
    df, na_list = nicaviz.reduce_mem_usage(df)

display(df.head())
display(df.nica.categorical_describe())
```

```python
categoricals = ['Subject Focus','Eyes','Face','Near','Action','Accessory',
                'Group','Collage','Human','Occlusion','Info','Blur']
continuous = ['Pawpularity']
```

```python
df.nica.mass_plot(
    plt_set = categoricals,
    columns = 3,
    plottype = "countplot")
```

```python
df.nica.mass_plot(
    plt_set = continuous,
    columns = 1,
    figsize = [7,5],
    plottype = "distplot")
```

```python
df.nica.pivot_plots(categoricals, 'Pawpularity', np.mean, palette=["Reds"])
```

```python
def pawplot(df, title):
    rows, cols = 4, 4
    fig, ax = plt.subplots(rows, cols, figsize = [rows*5,cols*5])

    for i, ax in zip(df["Id"].sample(rows*cols), ax.ravel()):
        ax.set(xticks=[], yticks=[])
        ax.set_title(f'Score: {df.loc[df["Id"] == i, "Pawpularity"].values[0]}')
        img = plt.imread(f'../input/petfinder-pawpularity-score/train/{i}.jpg')
        ax.imshow(img)

    fig.suptitle(title, fontsize=20, fontweight='bold')
    fig.tight_layout()
    fig.show()
```

```python
pawplot(df, "Random Selection")
```

```python
for bool_col in [
    'Subject Focus',
    'Eyes',
    'Face',
    'Near',
    'Action',
    'Accessory',
    'Group',
    'Collage',
    'Human',
    'Occlusion',
    'Info',
    'Blur']:
    pawplot(df.loc[df[bool_col] == 1], f'{bool_col} is True Selection')
```
