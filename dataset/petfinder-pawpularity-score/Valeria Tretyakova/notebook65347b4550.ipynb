```python
import numpy as np
import pandas as pd
from sklearn import ensemble
data_test = pd.read_csv('../input/petfinder-pawpularity-score/test.csv')
data_train = pd.read_csv('../input/petfinder-pawpularity-score/train.csv')

reg = ensemble.AdaBoostRegressor()
reg.fit(data_train.drop(['Pawpularity','Id'], 1), data_train['Pawpularity'])

sub = pd.read_csv('../input/petfinder-pawpularity-score/sample_submission.csv')
Pawpularity = reg.predict(data_test.drop('Id', 1))
sub['Pawpularity'] = Pawpularity
sub.to_csv('submission.csv', index = False)
```
