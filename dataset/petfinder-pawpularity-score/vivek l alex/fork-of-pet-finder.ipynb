### **Lets do some EDA**

```python
#no null values lets display some sample images
img_name=os.listdir(train_image)
random_images=random.choices(img_name,k=11)

fig=plt.figure(figsize=(60,60))
fig.suptitle('10 random images',fontsize=73)
for i in range(1,11):
    image=img.imread(os.path.join(train_image,random_images[i]))
    fi=fig.add_subplot(5,2,i)
    fi.set_title(random_images[i],fontsize=50)
    plt.imshow(image)  
    
plt.show()    
    
```

#### **We can show top highest pawpolarity images and least pawpolarity images**

```python
top_images=train_csv.sort_values(by='Pawpularity',ascending=False)
worst_images=train_csv.sort_values(by='Pawpularity',ascending=False)
top_images=top_images['Id'][:10]

fig=plt.figure(figsize=(60,60))
fig.suptitle('Top 10 pawpularity score',fontsize=73)
for i in range(0,10):
    image=img.imread(os.path.join(train_image,list(top_images)[i]+'.jpg'))
    fi=fig.add_subplot(5,2,i+1)
    fi.set_title(list(top_images)[i],fontsize=50)
    plt.imshow(image)  
    
plt.show()    
```

```python
worst_images=train_csv.sort_values(by='Pawpularity',ascending=True)
worst_images=worst_images['Id'][:10]
fig=plt.figure(figsize=(60,60))
fig.suptitle('Last 10 pawpularity score',fontsize=73)
for i in range(0,10):
    image=img.imread(os.path.join(train_image,list(worst_images)[i]+'.jpg'))
    fi=fig.add_subplot(5,2,i+1)
    fi.set_title(list(worst_images)[i],fontsize=50)
    plt.imshow(image)  
    
plt.show()    
```

```python
#plt.hist(train_csv['Pawpularity'],histtype='bar',color='' )
fig=plt.figure()
fig.suptitle('Data distribution of Pawpularity')
plt.axvline(train_csv['Pawpularity'].mean(),color='orange')
sns.histplot(x=train_csv['Pawpularity'],kde=True)
```

####    **Lets analyze images with more than 80 pawpolarity score and find what makes them different**

```python
images_80=train_csv[train_csv['Pawpularity']>80]
images_80
```

#### **Making count plots for each of the columns**

```python
images_80=train_csv[train_csv['Pawpularity']>80]
images_80
for col in images_80.columns:
        if col not in ['Id','Pawpularity']:
               print(images_80[col].value_counts())
               dicts=dict(images_80[col].value_counts())
               print(dicts.keys())
               plt.pie(list(dicts.values()),labels=list(dicts.keys()))
               plt.show()
                
                
```
