```python
# importing libraries

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
from tqdm import tqdm
import os
```

```python
# reading csv 
data = pd.read_csv('../input/petfinder-pawpularity-score/train.csv')
data.head()
```

```python
# visualising first 9 images from the given data
fig, ax = plt.subplots(3,3,figsize=(15,9))
for i, a in zip(data[['Id','Pawpularity']].head(9).iterrows(), ax.ravel()):
    a.set(xticks =[],yticks =[])
    img = plt.imread(f'../input/petfinder-pawpularity-score/train/{i[1][0]}.jpg')
    a.imshow(img)
    a.set_title(f'Id: {i[0]}, Pawpularity Score: {i[1][1]}',color='white')
fig.suptitle('Pawpularity Images',fontsize = 20,fontweight='bold')
fig.tight_layout()
fig.show()
```

```python
fig, ax = plt.subplots(12,3,figsize=(14,40))
for a in ax.ravel():
    a.set(xticks=[],yticks=[])
for r in range(12):
    label = data.columns[r+1]
    count = data[label].value_counts().sort_values()
    for i in [1,0]:
        img_id = data[data[label]==i].sample()['Id'].values[0]
        img =  plt.imread(f'../input/petfinder-pawpularity-score/train/{img_id}.jpg')
        c = 0 if i==1 else 2
        ax[r,c].imshow(img)
        ax[r,c].set_title(f'{label}={i}')
    ax[r, 1].pie(count, labels=[0, 1], autopct='%1.1f%%', wedgeprops = {'linewidth': 3})
    ax[r, 1].set_title(f'{label}', fontweight='bold', fontsize=20)
        
fig.tight_layout()
fig.show()
```

```python
# histogram for Pawpularity score

plt.figure(figsize=(12,5))
sns.histplot(data=data,x='Pawpularity',kde = True)
plt.axvline(data['Pawpularity'].mean(),c='purple',ls='-',lw=3)
plt.title('Pawpularity score Histogram',fontsize=10,fontweight = 'bold')
plt.show()
```

```python
# Pawpularity Heat map

plt.figure(figsize=(12,5))
sns.heatmap(data.corr(),linewidths=1,linecolor='black',annot = True)
plt.title('Pawpularity Heat map',fontsize=10,fontweight = 'bold')
plt.show()
```

```python
# 3 images with PAwpularity score 100
top = data[data['Pawpularity'] == 100]['Id']
fig ,ax = plt.subplots(1,3)
for i , ax in zip(top.sample(3),ax.ravel()):
    ax.set(xticks = [],yticks=[])
    img = plt.imread(f'../input/petfinder-pawpularity-score/train/{i}.jpg')
    ax.imshow(img)
fig.suptitle('top Paw score images',fontsize = 20,fontweight='bold')
fig.tight_layout()
fig.show()
```

```python
# 3 images with PAwpularity score 1 (low score)
bottom = data[data['Pawpularity'] == 1]['Id']
fig ,ax = plt.subplots(1,3)
for i , ax in zip(bottom.sample(3),ax.ravel()):
    ax.set(xticks = [],yticks=[])
    img = plt.imread(f'../input/petfinder-pawpularity-score/train/{i}.jpg')
    ax.imshow(img)
fig.suptitle('Bottom Paw score images',fontsize = 20,fontweight='bold')
fig.tight_layout()
fig.show()
```
