```python
# using facetgrid that is a great way to get information of our dataset
def get_FacetGrid(feature):

    # figure size
    plt.figure(figsize=(12,5))
    g = sns.FacetGrid(train, col=feature,size=5)
    g = g.map(sns.distplot, "Pawpularity")
    plt.show()
```
