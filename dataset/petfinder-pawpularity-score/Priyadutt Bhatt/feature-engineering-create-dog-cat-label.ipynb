```python
import numpy as np
import pandas as pd 
import os
import cv2
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.models import load_model
```

```python
train = pd.read_csv('../input/petfinder-pawpularity-score/train.csv')
test = pd.read_csv('../input/petfinder-pawpularity-score/test.csv')
```

```python
train["file_path"] = train["Id"].apply(lambda identifier: "../input/petfinder-pawpularity-score/train/" + identifier + ".jpg")
test["file_path"] = test["Id"].apply(lambda identifier: "../input/petfinder-pawpularity-score/test/" + identifier + ".jpg")
```

```python
def load_image(filename):
    # load the image
    img = load_img(filename, target_size=(224, 224))
    # convert to array
    img = img_to_array(img)
    # reshape into a single sample with 3 channels
    img = img.reshape(1, 224, 224, 3)
    # center pixel data
    img = img.astype('float32')
    img = img - [123.68, 116.779, 103.939]
    return img
```

```python
res=[]
def run_example_train():
    # load the image
    model = load_model('../input/catdogcnn/cats_vs_dogs_model.h5')
    for i in train['file_path'].values:
        img = load_image(i)
        # predict the class
        result = model.predict(img)
        result = result.flatten()
        res.append(round(result[0]))
```

```python
res_test=[]
def run_example_test():
    # load the image
    model = load_model('../input/catdogcnn/cats_vs_dogs_model.h5')
    for i in test['file_path'].values:
        img = load_image(i)
        # predict the class
        result = model.predict(img)
        result = result.flatten()
        res_test.append(round(result[0]))
```

```python
run_example_train()
train['class_label'] = res
```

```python
run_example_test()
test['class_label'] = res_test
```
