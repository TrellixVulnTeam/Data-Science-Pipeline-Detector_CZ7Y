```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
```

```python
train = pd.read_csv(r'../input/petfinder-pawpularity-score/train.csv')
test = pd.read_csv(r'../input/petfinder-pawpularity-score/test.csv')
```

```python
X = train.drop('Id', axis=1).iloc[:, :-1]
y = train.drop('Id', axis=1).iloc[:, -1]
```

```python
reg = RandomForestRegressor()
reg.fit(X, y)
```

```python
submission = pd.concat([test['Id'], pd.DataFrame(reg.predict(test.drop('Id', axis=1)), columns=['Pawpularity'])], axis=1)
```

```python
submission.to_csv('submission.csv')
```
