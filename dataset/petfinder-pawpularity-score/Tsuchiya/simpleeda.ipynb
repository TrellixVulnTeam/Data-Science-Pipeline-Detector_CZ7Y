```python
import cv2

import glob
from tqdm.notebook import tqdm

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

glob.glob('/kaggle/input/petfinder-pawpularity-score/*')
```

```python
base_path = '/kaggle/input/petfinder-pawpularity-score'
sample_submission = pd.read_csv(f'{base_path}/sample_submission.csv')
train = pd.read_csv(f'{base_path}/train.csv')
test = pd.read_csv(f'{base_path}/test.csv')

train_path = f'{base_path}/train'
test_path = f'{base_path}/test'
```

```python
print(len(train))
display(train.head())

print(len(test))
display(test.head())
```

```python
corr_df = train.drop(['Id'], axis=1).corr()

plt.figure(figsize=(10,10))
sns.heatmap(corr_df, linewidths=0.1, vmax=1.0, square=True, cmap=plt.cm.RdBu, annot=True)
plt.show()
```

```python
for c in train.drop(['Id', 'Pawpularity'], axis=1).columns:
    print(c, '+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++')
    for _, row in train[train[c]==1].head().iterrows():
        img = cv2.imread(f"{train_path}/{row['Id']}.jpg")
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        plt.imshow(img)
        plt.title(row['Pawpularity'])
        plt.show()
```

```python
img_size_list = []

x_max = 0
x_min = 9999
y_max = 0
y_min = 9999

for id_ in tqdm(train['Id']):
    img = cv2.imread(f"{train_path}/{id_}.jpg")
    y, x, _ = img.shape
    
    if x > x_max:
        x_max = x
        x_max_id = id_
    if x < x_min:
        x_min = x
        x_min_id = id_
    if y > y_max:
        y_max = y
        y_max_id = id_
    if y < y_min:
        y_min = y
        y_min_id = id_
    
    img_size_list.append([x, y])
```

```python
img = cv2.imread(f"{train_path}/{x_max_id}.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
print(img.shape)
plt.imshow(img)
plt.title('x_max_id')
plt.show()

img = cv2.imread(f"{train_path}/{x_min_id}.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
print(img.shape)
plt.imshow(img)
plt.title('x_min_id')
plt.show()

img = cv2.imread(f"{train_path}/{y_max_id}.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
print(img.shape)
plt.imshow(img)
plt.title('y_max_id')
plt.show()

img = cv2.imread(f"{train_path}/{y_min_id}.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
print(img.shape)
plt.imshow(img)
plt.title('y_min_id')
plt.show()
```

```python
img_size_df = pd.DataFrame(img_size_list, columns=['width', 'height'])

# X
plt.hist(img_size_df['width'])
plt.title('X')
plt.show()

# Y
plt.hist(img_size_df['height'])
plt.title('Y')
plt.show()
```

```python
train2 = pd.concat([train, img_size_df], axis=1)
```

```python
train2['rate'] = train2['width'] / train2['height']

plt.hist(train2['rate'])
plt.show()

img = cv2.imread(f"{train_path}/d61e49cd1ff16a8fe62f4f8e5c0fedff.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
print(img.shape)
plt.imshow(img)
plt.title('X << Y')
plt.show()

img = cv2.imread(f"{train_path}/ba6faa385160bedd384aed93329779fe.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
print(img.shape)
plt.imshow(img)
plt.title('X >> Y')
plt.show()
```
