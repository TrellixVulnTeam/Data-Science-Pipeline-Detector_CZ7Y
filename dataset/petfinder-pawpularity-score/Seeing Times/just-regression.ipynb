```python
train = pd.read_csv('/kaggle/input/petfinder-pawpularity-score/train.csv')
test = pd.read_csv('/kaggle/input/petfinder-pawpularity-score/test.csv')
train = train[(np.abs(stats.zscore(train['Pawpularity']))<1)]
```

```python
# define data
x = train.drop(columns='Id').iloc[:,:12]
y = train['Pawpularity']

# split data
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=210)

# modeling
liner = LinearRegression()
liner.fit(x_train,y_train)

# get coef,intercept
# liner.coef_
# liner.intercept_
y_pre = liner.predict(x_test)
test_pre = liner.predict(test.iloc[:,1:])
mse = mean_squared_error(y_test,y_pre)
rmse = np.sqrt(mse)
rmse
```
