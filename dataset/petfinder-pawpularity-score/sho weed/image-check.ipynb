# 4. class
### I'm sure there's a better way, but I did my best to make it work.

```python
class SHOW_IMG:
    def __init__(self, img_folder_path, train_table):
        self.img_folder_path = img_folder_path
        self.img_pathes = [img_folder_path+x+".jpg" for x in train_table["Id"]]
        self.lens = len(self.img_pathes)
        self.table = train_table
        self.my_tabel = train_table.loc[:,['Subject Focus', 'Eyes', 'Face', 'Near', 'Action', 'Accessory','Group', 'Collage', 'Human', 'Occlusion', 'Info', 'Blur']]
        
    def load_image(self, idx):
        img = img = Image.open(self.img_pathes[idx])
        return img
    
    def img_configuration(self, img, idx):
        idx = idx
        size = img.size
        df = self.my_tabel.iloc[idx,:]
        hit_columns = list(df[df==1].index) 
        Pawpularity = self.table.loc[idx,"Pawpularity"]
        text = f"Pawpularity:{Pawpularity}\nID : {idx}\nsize : {size}"
        text = text +"\n\n[HIT COLUMN]"
        for col in hit_columns:
            text = text +"\n"+ col
        #print(text)
        return text
    
    def show_image(self, imgs, rnd):
        plt.figure(figsize=(20,8))
        for i in range(3):
            text = self.img_configuration(imgs[i], rnd[i])
            size = imgs[i].size
            plt.subplot(1,3,i+1)
            plt.title(text)
            plt.imshow(imgs[i])
            plt.axis("off")
        plt.show()
    
    def START_random(self, N=3):
        print("-----Random_image_show-----")
        for i in range(N):
            print(f"-----{i+1}th-----")
            rnd = np.random.randint(0, self.lens, 3)
            imgs = [self.load_image(x) for x in rnd]
            self.show_image(imgs, rnd)
            
    def START_Pawpularity_1_100(self, N=3):
        print("-----10s_image_show-----")
        img_id = np.arange(self.lens)
        class_label = np.digitize(IMG.table["Pawpularity"], np.arange(0,120,10))
        for i in range(1,11):
            print(f"\n-----{i-1}s-----\n")
            for _ in range(N):
                idx = img_id[class_label==i]
                rnd = np.random.choice(idx, 3)
                imgs = [self.load_image(x) for x in rnd]
                self.show_image(imgs, rnd)
            
        print(f"\n-----100-----\n")
        for _ in range(N):
            idx = img_id[class_label==i]
            rnd = np.random.choice(idx, 3)
            imgs = [self.load_image(x) for x in rnd]
            self.show_image(imgs, rnd)
            
IMG = SHOW_IMG(train_dir, train_table)
```

# 7. Is there a secret to the image size?
#### It was never that simple....

```python
pathes = [x for x in  IMG.img_pathes]

size_hs, size_ws = [], []
for path in tqdm(pathes):
    img = Image.open(path)
    size_h, size_w, _ = np.array(img).shape
    size_hs.append(size_h)
    size_ws.append(size_w)

train_table["size_h"] = size_hs
train_table["size_w"] = size_ws
train_table["same_size"] = train_table["size_h"] == train_table["size_w"]
train_table["pixel"] = train_table["size_h"] * train_table["size_w"]
train_table
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(train_table.loc[:,["Pawpularity","size_h","size_w","pixel"]].corr(), 
            cmap="Blues", annot=True)
plt.show()
```
