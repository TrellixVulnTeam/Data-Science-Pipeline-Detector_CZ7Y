```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import pandas_profiling as pdp
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv("../input/petfinder-pawpularity-score/train.csv")
```

# check image distribution

```python
import cv2
import glob
```

```python
paths = glob.glob("../input/petfinder-pawpularity-score/train/*.jpg")
print(len(paths))
```

```python
# check image information
mean = []
std = []
h = []
w = []

for path in paths:
    img = cv2.imread(path)
    h.append(img.shape[0])
    w.append(img.shape[1])
    mean.append(img.mean())
    std.append(img.std())
```

```python
train["height"] = h
train["width"] = w
train["mean"] = mean
train["std"] = std
train["ratio_hw"] = train["height"]/train["width"]
train["diff_hw"] = abs(train["height"] - train["width"])
train
```

```python
train.to_csv("train.csv")
```

```python
pdp.ProfileReport(train)
```

```python
max_paths = "../input/petfinder-pawpularity-score/train/" + train[train.Pawpularity==100].Id.values + ".jpg"
```

## display image where the Pawpularity values is 100

```python
fig = plt.figure(figsize=(16,16*5))

for i,path in enumerate(max_paths):
    ax = fig.add_subplot(10, 2, i+1)
    img = cv2.imread(path)
    ax.imshow(img)
    if i==19:
        break
plt.show()
```

# Compare the value of Pawpularity 1 with 100 on images 

```python
min_paths = "../input/petfinder-pawpularity-score/train/" + train[train.Pawpularity==1].Id.values + ".jpg"
```

```python
fig = plt.figure(figsize=(16,16))
num = 0
for i in range(4):
    num += 1
    ax = fig.add_subplot(4, 2, num)
    img = cv2.imread(max_paths[i])
    ax.imshow(img)
    ax.set_title("pawpularity 100")
    num += 1
    ax = fig.add_subplot(4, 2, num)
    img = cv2.imread(min_paths[i])
    ax.imshow(img)
    ax.set_title("pawpularity 0")
plt.show()
```
