```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import cv2
```

```python
df = pd.read_csv("../input/simple-eda-using-pandas-profiling/train.csv")
df
```

```python
input_unique =df[["Subject Focus","Eyes","Face","Near","Action","Accessory","Group","Collage","Human","Occlusion","Info","Blur"]].drop_duplicates()
```

```python
input_unique["catesum"] = input_unique.apply(lambda x: x.sum(),axis=1)
```

```python
input_unique.sort_values("catesum")
```

```python
df["catesum"] = df[["Subject Focus","Eyes","Face","Near","Action","Accessory","Group","Collage","Human","Occlusion","Info","Blur"]].apply(lambda x: x.sum(),axis=1)
```

```python
maxcat_paths = "../input/petfinder-pawpularity-score/train/" + df[df.catesum==8].Id.values + ".jpg"
mincat_paths = "../input/petfinder-pawpularity-score/train/" + df[df.catesum==0].Id.values + ".jpg"
onecat_paths = "../input/petfinder-pawpularity-score/train/" + df[df.catesum==1].Id.values + ".jpg"
```

```python
df[df.catesum==8][["Subject Focus","Eyes","Face","Near","Action","Accessory","Group","Collage","Human","Occlusion","Info","Blur","Pawpularity"]]
```

```python
fig = plt.figure(figsize=(16,16))

for i,path in enumerate(maxcat_paths):
    ax = fig.add_subplot(2, 3, i+1)
    img = cv2.imread(path)
    ax.imshow(img)
    ax.set_title(f"Pawpularity:{df[df.catesum==8].Pawpularity.values[i]}")
plt.show()
```

```python
fig = plt.figure(figsize=(16,16))

for i,path in enumerate(mincat_paths):
    ax = fig.add_subplot(4, 4, i+1)
    img = cv2.imread(path)
    ax.imshow(img)
    ax.set_title(f"Pawpularity:{df[df.catesum==0].Pawpularity.values[i]}")
    if i==15:
        break
plt.show()
```
