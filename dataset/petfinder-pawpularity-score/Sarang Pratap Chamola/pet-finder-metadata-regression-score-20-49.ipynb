```python
#load in the data:
#source path (where the Pawpularity contest data resides)
path = '../input/petfinder-pawpularity-score/'

#Get the metadata (the .csv data) and put it into DataFrames
train_df = pd.read_csv(path + 'train.csv')

#show the dimensions of the train metadata.
print('train_df dimensions: ', train_df.shape)
train_df.head(3)
```

```python
test_df=pd.read_csv(path + 'test.csv')
test_df.head(3)
```

```python
train_df.skew()
```

```python
sns.histplot(train_df["Pawpularity"],kde=True)
```

```python
Target=train_df["Pawpularity"]
Train=train_df.drop(["Pawpularity",'Id'],axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(Train, Target, test_size = 0.2, random_state=5)
```

```python
X_train.head(2)
```

```python
model = CatBoostRegressor(iterations=100,
                          learning_rate=0.01,
                          depth=16,
                          loss_function='RMSE')
```

```python
#train the model
start = time.time()
model.fit(X_train,y_train)
stop = time.time()
```

```python
model_pred = model.predict(X_test)
print(f'Training time: {round((stop - start),3)} seconds')
model_RMSE = math.sqrt(mean_squared_error(y_test, model_pred))
print(f'tree_reg_RMSE: {round(model_RMSE,3)}')
```

```python
test_df = pd.read_csv('../input/petfinder-pawpularity-score/test.csv')
x_test_submission = test_df.drop(['Id'],axis=1) 
test_df['Pawpularity'] = model.predict(x_test_submission) 
submission_df = test_df[['Id','Pawpularity']]
submission_df.to_csv("submission.csv", index=False)
submission_df.head()
```
