```python
import cuml ,pickle
from cuml.svm import SVR
from cuml.linear_model import LogisticRegression, ElasticNet
from cuml.ensemble import RandomForestRegressor
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
train=pd.read_csv('../input/petfinder-pawpularity-score/train.csv')
train.head()
```

```python
X=train.drop(['Id','Pawpularity'],axis=1)
Y=train.Pawpularity.values
X=X.astype('float32')
X_train,X_valid,Y_train,Y_valid=train_test_split(X,Y,test_size=0.3,random_state=42)
```

# SVR

```python
model=SVR(C=20.0)
model.fit(X_train.values,Y_train)
```

# Logistic Regression

```python
model=LogisticRegression()
model.fit(X_train.values,Y_train)
```

# Elastic Net

```python
model=ElasticNet()
model.fit(X_train.values,Y_train)
```

# RandomForest

```python
model=RandomForestRegressor()
model.fit(X_train.values,Y_train)
```

# Rmse

```python
def Rmse(y_pred,y_val):
    return np.sqrt(((y_val-y_pred)**2).mean())
```

```python
pred=model.predict(X_valid)
rmse=Rmse(pred,Y_valid)

print(rmse)
```

```python
test=pd.read_csv("../input/petfinder-pawpularity-score/test.csv")
test=test.drop('Id',axis=1)
test=test.astype('float32')
pred=model.predict(test.values)
pred
```

```python
submission=pd.read_csv('../input/petfinder-pawpularity-score/sample_submission.csv')
submission['Pawpularity']=pred
submission.to_csv('submission.csv',index=False)
```
