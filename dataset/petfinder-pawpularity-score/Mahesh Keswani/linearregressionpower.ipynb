```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
from tqdm import tqdm
from sklearn.model_selection import KFold
import warnings
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score


warnings.filterwarnings("ignore")

trainMetadata = pd.read_csv('../input/petfinder-pawpularity-score/train.csv')
testMetadata = pd.read_csv('../input/petfinder-pawpularity-score/test.csv')
```

```python
X = trainMetadata.values[:, 1:-1]
Y = trainMetadata.values[: ,-1]
```

```python
from sklearn.linear_model import LinearRegression

# create model
model = LinearRegression()

# prepare the cross-validation procedure
cv = KFold(n_splits=10, random_state=1, shuffle=True)

# evaluate model
scores = cross_val_score(model, X, Y, scoring='neg_root_mean_squared_error', cv=cv, n_jobs=-1)

# report performance
print(np.mean(scores), np.std(scores))
```

```python
model = LinearRegression().fit(X, Y)

prediction = model.predict( testMetadata.values[:, 1:] )

submission = pd.DataFrame()
submission['Id'] = testMetadata['Id']
submission['Pawpularity'] = prediction
submission.to_csv('submission.csv',index = False)
```
