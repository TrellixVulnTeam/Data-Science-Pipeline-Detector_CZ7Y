# Why OOF

OOF stands for out-of fold. It uses for pre-summit summissions in cases of to check submission quality before use submission quota and to compare models quality.

```python
import pandas as pd
import numpy as np


from sklearn.model_selection import train_test_split


class Generator:
    def __init__(self, train_path, target, seed: int=1234, frac: float=0.8):
    
        self.df = pd.read_csv(train_path, index_col="Id")
        self.df = self.df.sample(frac=1, random_state=seed)
        
        split_index = int(self.df.shape[0]*frac)
        self.train = self.df[0: split_index]
        self.test = self.df[split_index:]
        
        print(f"Train target: {self.train.shape[0]}")
        print(f"Test target: {self.test.shape[0]}")
        mean = pd.concat([np.mean(self.train), np.mean(self.test)], axis=1)
        std = pd.concat([np.std(self.train), np.std(self.test)], axis=1)
        mean.columns = ["train", "test"]
        std.columns = ["train", "test"]
        print(f"Mean: {mean}")
        print(f"Std: {std}")
        

        
train_path = "../input/petfinder-pawpularity-score/train.csv"
target = "Pawpularity"

generator = Generator(train_path, target=target)

generator.train.to_csv("train.csv")
generator.test.to_csv("test.csv")
```
