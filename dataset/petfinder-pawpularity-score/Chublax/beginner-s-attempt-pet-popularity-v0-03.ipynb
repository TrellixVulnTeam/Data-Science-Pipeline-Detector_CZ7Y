```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor


ttrain_data = pd.read_csv("../input/petfinder-pawpularity-score/train.csv")
ttrain_data.head

train_data = ttrain_data.dropna(axis=0)

ttest_data = pd.read_csv("../input/petfinder-pawpularity-score/test.csv")
ttest_data.head()

test_data = ttest_data.dropna(axis=0)

submission_data = pd.read_csv("../input/petfinder-pawpularity-score/sample_submission.csv")

y = train_data.Pawpularity

features = ['Subject Focus', 'Eyes', 'Face', 'Near', 'Action', 'Accessory','Group', 'Collage','Occlusion', 'Human','Info','Blur']
X = pd.get_dummies(train_data[features])
X_train = train_data[features]

model = RandomForestClassifier(n_estimators=33,max_depth=2, random_state=1)
model.fit(X, y)
model_predictions = model.predict(X)
model_mae = mean_absolute_error(model_predictions, y)
print("model MAE for Random Forest Model: {}".format(model_mae))

# Define the model. Set random_state to 1
rf_model = DecisionTreeRegressor(max_leaf_nodes=1000, random_state=1)
rf_model.fit(X_train, y)
rf_val_predictions = rf_model.predict(X)
rf_val_mae = mean_absolute_error(rf_val_predictions, y)
print("rf MAE for Random Forest Model: {}".format(rf_val_mae))


predictions = rf_model.predict(X)
output = pd.DataFrame({'Id': train_data.Id, 'Pawpularity': predictions})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")


print("First in-sample predictions:", model.predict(X.head()))
print("First rf in-sample predictions:", rf_model.predict(X.head()))
print("Actual target values for those homes:", y.head().tolist())


```
