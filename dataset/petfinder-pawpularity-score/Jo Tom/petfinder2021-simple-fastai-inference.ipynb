# PetFinder2021 simple Fastai Inference

This is a simple Fastai Inference Notebook. The models are trained on images only.

The Train notebook can be found [here](https://www.kaggle.com/joatom/petfinder2021-fastai-train).

Related Discussion [here](https://www.kaggle.com/c/petfinder-pawpularity-score/discussion/294822#1617055).


```python
import pandas as pd
import numpy as np
import fastai
from fastai.vision.all import *
# You can write up to 20GB to the current directory (/kaggle/working/) that gets preserved as output when you create a version using "Save & Run All" 
# You can also write temporary files to /kaggle/temp/, but they won't be saved outside of the current session
```

```python
submission = pd.read_csv('../input/petfinder-pawpularity-score/sample_submission.csv')

test = submission.copy()

test['Id'] = 'test/' + test['Id'] + '.jpg'
test.head()
```

```python
def inf_run(fold):
    learn = load_learner(fname = Path(f'../input/petfinder2021-fastai-train/export_fold_{fold}.pkl'), cpu=False)
    test_dl = learn.dls.test_dl(test)
    preds, _ = learn.get_preds(dl=test_dl) # tta
    
    return preds
```

```python
submission['Pawpularity'] = 0

for fold in range(7):
    preds = inf_run(fold)
    test[f'preds_{fold}'] = preds.numpy()[:,0]
```

```python
test.head()
```

```python
submission.Pawpularity = test[['preds_'+str(i) for i in range(7)]].mean(axis=1)
submission.to_csv('submission.csv', index = False)

submission.head()
```
