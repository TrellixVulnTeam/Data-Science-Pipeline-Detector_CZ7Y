```python
model = new_model = tf.keras.models.load_model('../input/pet-pawpularity-model/my_model.h5')
```

```python
test_csv = pd.read_csv("../input/petfinder-pawpularity-score/test.csv") 
test_imgs_path = '../input/petfinder-pawpularity-score/test'
test_csv['image_path'] = test_csv['Id'].apply(lambda x: f"{test_imgs_path}/{x}.jpg")
test_imgs = np.array([cv2.resize(cv2.imread(file),(224,224)) for file in test_csv['image_path']])
    
prediction = model.predict([test_imgs,test_csv.drop(columns=['Id','image_path'])])
count = 0
data = []
for i in test_csv.itertuples():
  data.append([i[1],prediction[count][0]])
  count+=1
prediction_df = pd.DataFrame(data,columns=['Id','Pawpularity'])
prediction_df.to_csv('submission.csv',header=True,index=None)
prediction_df
```
