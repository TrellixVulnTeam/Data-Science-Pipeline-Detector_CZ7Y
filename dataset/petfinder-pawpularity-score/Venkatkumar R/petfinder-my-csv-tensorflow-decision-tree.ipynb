```python
#mathematical analysis
import numpy as np
import pandas as pd
from pathlib import Path
import os.path
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import os
import cv2
#preprocessing
from sklearn import model_selection
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
from sklearn import preprocessing
#Model
from sklearn.metrics import mean_squared_error
```

## **Import data**

```python
#create file path
train_image = '../input/petfinder-pawpularity-score/train'
test_image = '../input/petfinder-pawpularity-score/test'

train_data = pd.read_csv('../input/petfinder-pawpularity-score/train.csv')
test_data = pd.read_csv('../input/petfinder-pawpularity-score/test.csv')
sample = pd.read_csv('../input/petfinder-pawpularity-score/sample_submission.csv')
print('Shape of input data :',train_data.shape,test_data.shape,sample.shape)
print('Identify the Null values : ',train_data.isnull().sum())
train_data.sample(2)
```

```python

import tensorflow_decision_forests as tfdf

train_tfds = tfdf.keras.pd_dataframe_to_tf_dataset(train_data,label="Pawpularity")
test_tfds = tfdf.keras.pd_dataframe_to_tf_dataset(test_data)

#Build the model
model = tfdf.keras.RandomForestModel()
model.fit(train_tfds)

#Model summary
model.summary()

pred = model.predict(test_tfds)

#prediction output
sample['Pawpularity'] = pred
sample.to_csv('submission.csv',index= False)
print('Success')
```
