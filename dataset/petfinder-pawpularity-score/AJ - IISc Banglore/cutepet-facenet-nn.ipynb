```python
def img_to_encoding(array, model):
    img = np.around(array / 255.0, decimals=12)
    x_train = np.expand_dims(img, axis=0)
    embedding = model.predict_on_batch(x_train)
    return embedding / np.linalg.norm(embedding, ord=2)
```

```python
T=[]
for i in range(Train.shape[0]):
    T.append(img_to_encoding(Train[i], model))
```

```python
from keras.models import Sequential
from keras.layers import Dense,BatchNormalization,Dropout
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
import tensorflow as tf
```

```python
model = Sequential()
model.add(Dense(16, input_dim=137, kernel_initializer='normal', activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(32, kernel_initializer='normal', activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.2))
model.add(Dense(64, kernel_initializer='normal',activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(128, kernel_initializer='normal', activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.2))
model.add(Dense(256, kernel_initializer='normal',activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(512, kernel_initializer='normal', activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.2))
model.add(Dense(128, kernel_initializer='normal',activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(64, kernel_initializer='normal', activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(8, kernel_initializer='normal', activation='relu'))
model.add(Dense(1, kernel_initializer='normal'))
    # Compile mode
model.compile(optimizer='adam',
              loss='mean_squared_error',
              metrics=['accuracy'])
early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_accuracy', patience=30, restore_best_weights=True)
```

```python
X_train, X_val, Y_train, Y_val = train_test_split(AJ, Y, test_size=0.3, random_state=1)
```

```python
Train_data =X_train#np.expand_dims(X_train, axis=-1)
Valid_data=X_val#np.expand_dims(X_val, axis=-1)
Train_abel=Y_train
Valid_label=Y_val
Valid_label= np.array(Valid_label)
Train_abel = np.array(Train_abel)
Valid_label=Valid_label.reshape(-1)
Train_abel=Train_abel.reshape(-1)
```

```python
tf.config.experimental_run_functions_eagerly(True)
history = model.fit(Train_data, Train_abel, epochs=100,batch_size=100, 
                    validation_data=(Valid_data, Valid_label),callbacks=[early_stopping])
```
