### This is a bare-bones submission with just a simple RF model trained on the photo metadata features
No hyperparameter tuning, no training on the image data etc
This can be used by anyone who wants to try a sample submission for this competition, or even their very first competition submission on Kaggle

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# import random forest regressor
from sklearn.ensemble import RandomForestRegressor

# Input data files are available in the read-only "../input/" directory

# You can write up to 20GB to the current directory (/kaggle/working/) that gets preserved as output when you create a version using "Save & Run All" 
# You can also write temporary files to /kaggle/temp/, but they won't be saved outside of the current session
```

```python
# train and test CSV contain the photo metadata features 
#binary flags on criteria such as 'eyes visible?', 'blurred?', 'is the pet doing an action'

train = pd.read_csv('../input/petfinder-pawpularity-score/train.csv')
train.head(2)
```

```python
train.drop(columns=['Id','Pawpularity'])
```

```python
train['Pawpularity']
```

```python
model_rf = RandomForestRegressor()
```

```python
model_rf.fit(X=train.drop(columns=['Id','Pawpularity']),y=train['Pawpularity'])
```

```python
# test with example test file to generate submission.csv
test = pd.read_csv('../input/petfinder-pawpularity-score/test.csv')
test.head(2)
```

```python
 test_preds = model_rf.predict(test.drop(columns=['Id']))
```

```python
test_preds
```

```python
sample_submission = pd.read_csv('../input/petfinder-pawpularity-score/sample_submission.csv')
```

```python
sample_submission.head(2)
```

## These are the same IDs as the test.csv, so we just need to update the Pawpularity column of the sample_submission file provided, with our predictions on the test set instead

```python
sample_submission.Pawpularity = test_preds
sample_submission.to_csv("submission.csv", index=False)
```
