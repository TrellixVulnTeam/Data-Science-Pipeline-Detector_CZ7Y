Baseline stratkfld with lgb regressor with only meta features (that's right, no images) 

Best RMSE on public lb: 20.50347

```python
# data manipulation
import pandas as pd

# cross validation
from sklearn.model_selection import StratifiedKFold

# metrics for model performance with RMSE
from sklearn.metrics import mean_squared_error
```

```python
class args:
    rand = 23
    train_path = "../input/petfinder-pawpularity-score/train.csv"
    test_path = "../input/petfinder-pawpularity-score/test.csv"
    submission_path = "../input/petfinder-pawpularity-score/sample_submission.csv"
    target = 'Pawpularity'
    splits = 5
```

```python
skf = StratifiedKFold(5, shuffle=True,random_state= args.rand)
```

```python
train = pd.read_csv(args.train_path, index_col=0)
```

```python
X = train.drop(args.target, axis=1)
y = train[args.target]
```

```python
import lightgbm as lgb
```

```python
model = lgb.LGBMRegressor(random_state=args.rand)
```

## Model training

```python
rmse = 0
for fld, (train_index, test_index) in enumerate(skf.split(X, y)):
    X_train = X.iloc[train_index]
    y_train = y.iloc[train_index]
    X_test = X.iloc[test_index]
    y_test = y.iloc[test_index]
    
    model.fit(X_train, y_train)
    
    predict = model.predict(X_test)
    
    curr_rmse = mean_squared_error(y_test, predict, squared=False)
    rmse += curr_rmse
    print(f"RMSE for fold:{fld} is {curr_rmse}.")
print(f"Average RMSE is {rmse / args.splits}")
```

```python
model.fit(X, y)
```

## Test data prediction

```python
test = pd.read_csv(args.test_path, index_col=0)
```

```python
predict = model.predict(test)
```

## Submission

```python
submission = pd.read_csv(args.submission_path, index_col=0)
submission[args.target] = predict
submission.to_csv('submission.csv', index=True)
```
