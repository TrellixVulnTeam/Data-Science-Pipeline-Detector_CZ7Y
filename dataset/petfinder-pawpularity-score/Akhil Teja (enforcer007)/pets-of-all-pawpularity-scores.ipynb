### Hidden Code for displaying Pets

```python
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
import os

data_folder = Path("../input/petfinder-pawpularity-score")
meta_data_train = data_folder/"train.csv"
images_train = data_folder/"train"

df = pd.read_csv(meta_data_train)
df = df.set_index("Id")
df.columns = [i.lower() for i in df.columns]

class SubMapping:
    def __init__(self):
        self._id2object = {}
    def from_strings(self,strings,func):
        ids = list(map(func,strings))
        self.from_lists(ids,strings)
        return self
    def from_lists(self,list1,list2):
        self._id2object = {i:j for i,j in zip(list1,list2)}
        return self
    def __getitem__(self,id):
        return self._id2object[id]
    def __len__(self):
        return len(self._id2object)
    def __iter__(self):
        for i in self._id2object:
            yield i,self[i]
    def values(self):
        return list(zip(self._id2object.keys(),self._id2object.values()))
    
    def rand_values(self,n):
        return random.choices(population=self.values(), k=n)
    
class FileIdMapping(SubMapping):
    def from_folderpath(self,folderpath,func):
        files = os.listdir(folderpath)
        ids = map(func,files)
        full_path_files = map(lambda x: os.path.join(folderpath,x),files)
        return self.from_lists(ids,full_path_files)

def plot_images(images,n_cols=3,size=(12,12),titles=None):
    n = len(images)
    n_rows = n//n_cols+bool(n%n_cols)
    
    fig,axs = plt.subplots(n_rows,n_cols,figsize=size)
    if isinstance(titles, (list,pd.Series,tuple)):
        if not len(titles):
            titles = n*["No-Title"]
    else:
        raise ValueError("Titles should be iterator")
    for ax,img,title in zip(axs.flatten(),images,titles):
        image = plt.imread(img)
        ax.imshow(image)
        ax.set_title("Pawpularity Score: {}".format(title))

fm = FileIdMapping()
fm = fm.from_folderpath(images_train,lambda x: x.split(".")[0])

def random_plot_images(n=30):
    image_data = fm.rand_values(n)
    images_ids = list(map(lambda x:x[0],image_data))
    images_pawpularity = df.loc[images_ids]["pawpularity"].sort_values()
    images_paths = [fm[i] for i in images_pawpularity.index]
    plot_images(images_paths,titles=images_pawpularity,size=(20,60))
```

### Beautiful Pets sorted by <u>increasing</u> pawpularity score
