```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('../input/petfinder-pawpularity-score/train.csv')
test = pd.read_csv('../input/petfinder-pawpularity-score/test.csv')
```

```python
train.drop('Id', axis=1, inplace=True)
test.drop('Id', axis=1, inplace=True)
```

```python
train.head()
```

```python
y, X = train['Pawpularity'], train.drop('Pawpularity', axis=1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
corrmat = train.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(13, 13))
g = sns.heatmap(train[top_corr_features].corr(), annot=True, cmap="RdYlGn")
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
model = GradientBoostingRegressor(n_estimators=5, max_depth=3, learning_rate=0.9)
```

```python
model.fit(X, y)
```

```python
prediction = model.predict(test)
```

```python
solution = pd.read_csv('../input/petfinder-pawpularity-score/sample_submission.csv')
```

```python
solution.head()
```

```python
solution['Pawpularity'] = prediction
```

```python
solution.head()
```

```python
solution.to_csv('submission.csv', index=False)
```
