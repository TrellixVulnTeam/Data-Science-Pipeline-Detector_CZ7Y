```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv("../input/petfinder-pawpularity-score/train.csv")
test = pd.read_csv("../input/petfinder-pawpularity-score/test.csv")
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(train.drop(['Id', 'Pawpularity'], axis=1), 
                                                    train['Pawpularity'], 
                                                    test_size=0.2, 
                                                    random_state=612)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
# RandomForest
rf = RandomForestRegressor(n_estimators=200, max_depth=5)
rf.fit(x_train, y_train)
pred = rf.predict(x_test)
np.sqrt(mean_squared_error(y_test, pred))
```

```python
sub = pd.read_csv("../input/petfinder-pawpularity-score/sample_submission.csv")

sub['Pawpularity'] = rf.predict(test.drop(['Id'], axis=1))
sub.to_csv("submission.csv", index=False)
sub
```
