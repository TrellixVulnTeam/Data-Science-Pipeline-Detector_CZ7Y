```python
import numpy as np
import pandas as pd
import os
from sklearn.model_selection import *
import warnings
warnings.filterwarnings('ignore')
```

```python
train = pd.read_csv('../input/petfinder-pawpularity-score/train.csv')
```

```python
# Simple KFold - 5 folds

kfold = KFold(n_splits=5, shuffle=True, random_state=42)
print('Train Mean | Train Std | Valid Mean | Valid Std')
print('-'*48)
for i, (t, v) in enumerate(kfold.split(train)):
    print(f'{train.Pawpularity.iloc[t].mean():.7f} | {train.Pawpularity.iloc[t].std():.6f} | {train.Pawpularity.iloc[v].mean():.7f} | {train.Pawpularity.iloc[v].std():.7f}')
```

```python
# StratifiedKFold no bins - 5 folds

kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
print('Train Mean | Train Std | Valid Mean | Valid Std')
print('-'*48)
for i, (t, v) in enumerate(kfold.split(train, train.Pawpularity)):
    print(f'{train.Pawpularity.iloc[t].mean():.7f} | {train.Pawpularity.iloc[t].std():.6f} | {train.Pawpularity.iloc[v].mean():.7f} | {train.Pawpularity.iloc[v].std():.7f}')
```

```python
# StratifiedKFold w/ bins (Sturge's Rule) - 5 folds

kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
bins = pd.cut(train.Pawpularity, bins=int(1 + np.log2(len(train))), labels=False)
print('Train Mean | Train Std | Valid Mean | Valid Std')
print('-'*48)
for i, (t, v) in enumerate(kfold.split(train, bins)):
    print(f'{train.Pawpularity.iloc[t].mean():.7f} | {train.Pawpularity.iloc[t].std():.6f} | {train.Pawpularity.iloc[v].mean():.7f} | {train.Pawpularity.iloc[v].std():.7f}')
```

```python
# StratifiedKFold w/ bins (quantiles) - 5 folds

kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
bins = pd.qcut(train.Pawpularity, q=50, labels=False, duplicates='drop')
print('Train Mean | Train Std | Valid Mean | Valid Std')
print('-'*48)
for i, (t, v) in enumerate(kfold.split(train, bins)):
    print(f'{train.Pawpularity.iloc[t].mean():.7f} | {train.Pawpularity.iloc[t].std():.6f} | {train.Pawpularity.iloc[v].mean():.7f} | {train.Pawpularity.iloc[v].std():.7f}')
```
