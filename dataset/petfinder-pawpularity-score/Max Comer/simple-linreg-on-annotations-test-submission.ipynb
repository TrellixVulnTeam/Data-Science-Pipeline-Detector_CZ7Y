```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
import math
```

### Load and process the annotations data only

```python
data_dir = '../input/petfinder-pawpularity-score/'
```

```python
df = pd.read_csv(f'{data_dir}train.csv')
df.head()
```

```python
y = np.array(df.iloc[:, 13])
y
```

```python
X = np.array(df.iloc[:, 1:13])
X
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size = 0.2, random_state=13)
```

### Train a simple linear regression

```python
reg = LinearRegression().fit(X_train, y_train)
```

```python
reg.coef_
```

```python
# Training RMSE
y_train_pred = reg.predict(X_train)
math.sqrt(mean_squared_error(y_train, y_train_pred))
```

```python
# Validation RMSE
y_valid_pred = reg.predict(X_valid)
math.sqrt(mean_squared_error(y_valid, y_valid_pred))
```

```python
# Check a Very simple MSE from using the mean.  
avg = np.ones(len(df))*33
print(avg)
math.sqrt(mean_squared_error(y, avg))
```

### Output predictions on the test set

```python
test_df = pd.read_csv(f'{data_dir}test.csv')
img_names = list( test_df.iloc[:, 0].values)
X_test = np.array(test_df.iloc[:, 1:13])
test_df.head()
```

```python
outputs = reg.predict(X_test)
```

```python
output_df = pd.DataFrame({'Id': img_names, 'Pawpularity': outputs.round(decimals=2)})
```

```python
# Write the output in the required format
output_df.to_csv('submission.csv', index=False)
```
