```python
import tensorflow as tf

from tensorflow import keras

meta_train = pd.read_csv("../input/petfinder-pawpularity-score/train.csv")

labels = meta_train.keys()[1:-1]
X = meta_train[labels]
y = meta_train.Pawpularity


#train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    
#)

#images = []
#for dirname, _, filenames in os.walk("/kaggle/input/petfinder-pawpularity-score/train/"):
#    for filename in filenames:
#        images.append(cv2.imread(dirname+filename))
    
#images

#img = cv2.imread("../input/petfinder-pawpularity-score/train/0007de18844b0dbbb5e1f607da0606e0.jpg")
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

rf_reg = RandomForestRegressor(random_state=1)
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=1)

rf_reg.fit(X_train, y_train)

y_pred = rf_reg.predict(X_val)

preds = mean_squared_error(y_val, y_pred, squared=False)

print(preds)
```

```python
test = pd.read_csv("../input/petfinder-pawpularity-score/test.csv")

X_test = test[test.keys()[1:]]
y_pred = rf_reg.predict(X_test)

sub = pd.DataFrame({'Id':test.Id, "Pawpularity":y_pred})
sub.to_csv("submission.csv", index=False)

print(sub)
```
