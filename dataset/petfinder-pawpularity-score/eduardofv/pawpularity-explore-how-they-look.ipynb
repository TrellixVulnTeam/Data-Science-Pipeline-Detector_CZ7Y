```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import tensorflow.keras as keras
import matplotlib.pyplot as plt

train = pd.read_csv("../input/petfinder-pawpularity-score/train.csv")

def show(img_id):
    img = keras.preprocessing.image.load_img(f"{IMG_DIR}/{img_id}.jpg") 
    plt.imshow(img)
    plt.axis('off')
    plt.show()
```

```python
print(train.Pawpularity.describe())
train.Pawpularity.hist(bins=20)
```

## These are Pawpular:

```python
pawpular = train[train.Pawpularity >= 90]
pawpular.sample(10).Id.apply(show)
```

## While these are not-so-pawpular:

```python
unpawpular = train[train.Pawpularity <= 10]
unpawpular.sample(10).Id.apply(show)
```
