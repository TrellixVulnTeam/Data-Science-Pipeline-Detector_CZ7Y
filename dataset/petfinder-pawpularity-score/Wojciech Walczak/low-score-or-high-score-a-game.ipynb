```python
import random
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

```python
df = pd.read_csv('/kaggle/input/petfinder-pawpularity-score/train.csv')
```

```python
SAMPLE_SIZE=5
sample = [(k, 0) for k in random.sample(df[df['Pawpularity'] <= 10]['Id'].values.tolist(), SAMPLE_SIZE)]
sample.extend([(k, 1) for k in random.sample(df[df['Pawpularity'] > 90]['Id'].values.tolist(), SAMPLE_SIZE)])
random.shuffle(sample)
```

```python
total = 0
for id, score in sample:
    img = mpimg.imread(f'/kaggle/input/petfinder-pawpularity-score/train/{id}.jpg')
    plt.imshow(img)
    plt.show()
    value = int(input('Type 0 for low score and 1 for high score: '))
    if value == score:
        total += 1
        print("correct!")
    else:
        print("wrong!")
print(f"Final score: {total}/{2*SAMPLE_SIZE}")
```
