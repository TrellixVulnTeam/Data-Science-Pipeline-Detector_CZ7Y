**Evaluation Metric**

Metric for Evaluation
 Root Mean Square Error(RMSE) is used for evaluation of results.
 
RMSE is defined as


$$\sqrt{\Sigma_{i=1}^{n}{\Big(\frac{\hat{y}_i - y_i}{n}\Big)^2}}$$

where $n$ denotes the number of samples, $y_i$ the ground truth value and $\hat{y}_i$ the prediction value.

```python
updated_df=pd.merge(train_csv, new_df, on='img_path')
```

```python
# reading all the images and its meta data with target value
# target list contains the target value
#train_data contains both img data and metadata


train=[]
y=[]
path="../input/petfinder-pawpularity-score/train"

for i in range(len(updated_df)):
    train1=updated_df.drop(['Id','img_path','Pawpularity'],axis=1)[updated_df.index==i].values
    
    y.append(updated_df.Pawpularity[i])
    path=updated_df.img_path[i]
    img1=cv2.imread(path)
    img1=cv2.resize(img1,(128,128))
    img1=cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
    
    train.append([img1,train1[0]])
    
```

```python
from sklearn.model_selection import train_test_split
X_train,X_valid,y_train,y_valid=train_test_split(train,y,test_size=0.1,random_state=42,shuffle=True)
```

```python
from sklearn.model_selection import train_test_split
X_valid1,X_valid2,y_valid1,y_valid2=train_test_split(X_valid,y_valid,test_size=0.5,random_state=42,shuffle=True)
```

```python
#splitting X_traininto two different lists
train_img_data=[]
train_feature_data=[]
for i,j in X_train:
    train_img_data.append(i)
    train_feature_data.append(j)
```

```python
valid_img_data1=[]
valid_feature_data1=[]
for i,j in X_valid1:
    valid_img_data1.append(i)
    valid_feature_data1.append(j)

valid_img_data2=[]
valid_feature_data2=[]
for i,j in X_valid2:
    valid_img_data2.append(i)
    valid_feature_data2.append(j)
```

```python
train_feature_data=np.array(train_feature_data)
train_feature_data.shape

valid_feature_data1=np.array(valid_feature_data1)
valid_feature_data1.shape

valid_feature_data2=np.array(valid_feature_data2)
valid_feature_data2.shape
```

```python
# normalising img data by dividing it by 255 the result matrix will have value between 0 and 1
train_img_data=np.array(train_img_data)
train_img_data.shape

valid_img_data1=np.array(valid_img_data1)
valid_img_data1.shape

valid_img_data2=np.array(valid_img_data2)
valid_img_data2.shape
```

```python

y_train=np.array(y_train)

y_valid1=np.array(y_valid1)

y_valid2=np.array(y_valid2)
```

```python
def clipvalues(values):
    for i in range(len(values)):
        if values[i]<1:
            values[i]=1
        if values[i]>100:
            values[i]=100
    return values
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential,Model
from tensorflow.keras.layers import Flatten,BatchNormalization,Convolution2D,MaxPooling2D,Dropout,Dense,Concatenate

from tensorflow.keras import Input
from tensorflow.keras.callbacks import EarlyStopping,ReduceLROnPlateau
```

```python
#importing Xception model from keras.application.

from tensorflow.keras.applications import xception

# Assigining the weights
xception = xception.Xception(weights='../input/xception-weights/xception_weights_tf_dim_ordering_tf_kernels_notop.h5',include_top=False,input_shape=(128, 128, 3))
```

```python
len(xception.layers)
```

```python
import sklearn
for layer in xception.layers[:120]:
        layer.trainable=False
first= Flatten()(xception.output)
first= Dense(1024,activation="relu")(first)
#first=BatchNormalization()(first)
first=Dropout(0.8)(first)

first= Dense(1024,activation="relu")(first)
#first=BatchNormalization()(first)
first=Dropout(0.8)(first)

tabular_input = Input(shape=(15,))
second=Dense(1024,activation="relu")(tabular_input)
#second=BatchNormalization()(second)
second=Dropout(0.8)(second)

second=Dense(512,activation="relu")(second)
#second=BatchNormalization()(second)
second=Dropout(0.8)(second)

combined = Concatenate(axis=1)([first, second])
result=Dense(512,activation="relu")(combined)
#result=BatchNormalization()(result)
result=Dropout(0.8)(result)

result=Dense(256,activation="relu")(combined)
#result=BatchNormalization()(result)
result=Dropout(0.8)(result)

result=Dense(1,activation="relu")(result)
    
    
image_input = Input(shape=(128,128,3))
model = tf.keras.Model(inputs=[xception.input, tabular_input], outputs=[result])
    
early_stopping = EarlyStopping(patience = 100)
reduce_lr = ReduceLROnPlateau(factor=0.1, patience=60,min_lr=1e-9)

model.compile(loss=tf.keras.losses.MeanSquaredError(), optimizer='adam', metrics=[tf.keras.metrics.RootMeanSquaredError(name="rmse"), "mae", "mape"])
predictor=model.fit((train_img_data,train_feature_data),y_train,validation_split=0.2, epochs=250,batch_size=32, callbacks=[early_stopping,reduce_lr])
    
y_pred=model.predict((valid_img_data1,valid_feature_data1))
y_pred2=model.predict((valid_img_data2,valid_feature_data2))

y_pred=clipvalues(y_pred)
y_pred2=clipvalues(y_pred2)

y_pred=y_pred.reshape(-1,1)
y_pred2=y_pred2.reshape(-1,1)


y_valid1=y_valid1.reshape(-1,1)
y_valid2=y_valid2.reshape(-1,1)
    
mse1 = sklearn.metrics.mean_squared_error(y_valid1,y_pred)
mse2 = sklearn.metrics.mean_squared_error(y_valid2,y_pred2)
rmse0 = math.sqrt(mse1)
rmse1 = math.sqrt(mse2)
```

```python
print("RMSE on predicted data with first validation set= ",rmse0)
print("RMSE on predicted data with second validation set= ",rmse1)
```
