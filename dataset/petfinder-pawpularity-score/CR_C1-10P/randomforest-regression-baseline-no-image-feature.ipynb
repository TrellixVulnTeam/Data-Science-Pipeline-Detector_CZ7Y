## RandomForest Regression 1st Baseline

Just putting my first initial baseline submission with random forest regression with no Image features.

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
```

```python
train_df = pd.read_csv("../input/petfinder-pawpularity-score/train.csv")
trainX = train_df.iloc[: ,1: -1]
trainY = train_df.iloc[:,-1]
trainX.head()
```

```python
test_df = pd.read_csv("../input/petfinder-pawpularity-score/test.csv")
testX = test_df.iloc[: ,1: ]
testX.head()
```

```python
test_df.columns
```

```python
regr = RandomForestRegressor(max_depth=2, random_state=0)
regr.fit(trainX, trainY)
```

```python
predictions = regr.predict(testX)
predictions
```

```python
res = []
for idx in range(len(test_df["Id"])):
    res.append((test_df["Id"][idx], predictions[idx]))
    

```

```python
submision_df = pd.DataFrame(res, columns=["Id", "Pawpularity"])
submision_df.to_csv("submission.csv", index=False)
```
