```python
import numpy as np
import pandas as pd
import joblib 
```

```python
MODEL_PATH = "../input/pefinder-xgbm-model/petfinder_xgbm.pch"
NEW_TEST_PATH = "../input/pefinder-xgbm-model/new_test.csv"
ID = "Id"
TARGET = "Pawpularity"

SAMPLE_SUBMISSION_PATH = "../input/petfinder-pawpularity-score/sample_submission.csv"
SUBMISSION_PATH = "submission.csv"
```

```python
model = joblib.load(MODEL_PATH) 
```

```python
model
```

```python
new_test = pd.read_csv(NEW_TEST_PATH)
new_test
```

```python
X_test = new_test.drop([ID],axis=1)
X_test[:5]
```

```python
pred_test = model.predict(X_test)
```

```python
sub = pd.read_csv(SAMPLE_SUBMISSION_PATH)
sub[TARGET] = pred_test
sub.to_csv(SUBMISSION_PATH)
sub.head()
```
