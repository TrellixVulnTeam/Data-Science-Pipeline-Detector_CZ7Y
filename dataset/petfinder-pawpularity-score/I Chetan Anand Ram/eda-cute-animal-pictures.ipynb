# I love animals. This notebook is made just for viewing cute animals.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def plot_pictures(target_df):
    plt.figure(figsize=(20, 50))
    
    n_rows = min(60, target_df.shape[0])
    
    for i in range(n_rows):
        row = target_df.iloc[i]
        img_path = f"../input/petfinder-pawpularity-score/train/{row['Id']}.jpg"
        Pawpularity = row["Pawpularity"]
        img = cv2.cvtColor(cv2.imread(img_path), cv2.COLOR_BGR2RGB)
        plt.subplot(12, 5, i+1)
        plt.title(f"Pawpularity: {Pawpularity}")
        plt.imshow(img)
    plt.tight_layout()
    plt.show()
    plt.close()
```

```python
train_df = pd.read_csv('../input/petfinder-pawpularity-score/train.csv')
train_df.head(2)
```

```python
plt.figure()
plt.hist(train_df["Pawpularity"], bins=100)
plt.show()
```

# Pawpularity < 10

```python
target_df = train_df[train_df["Pawpularity"] <= 10]
plot_pictures(target_df)
```

```python
target_df = train_df[(10 < train_df["Pawpularity"]) & (train_df["Pawpularity"] <= 20)]
plot_pictures(target_df)
```
