As mentionned in my [previous notebook](https://www.kaggle.com/maximeperez/first-analysis-of-the-data), data seems not to be very pertinent and i was wondering what score could i expect returning always the same value.

So i decide to return for each row of the test dataset the mean pawpularity score of the training dataset.

```python
import pandas as pd
```

```python
training_data = pd.read_csv("../input/petfinder-pawpularity-score/train.csv")
mean = training_data["Pawpularity"].mean()
```

```python
submission = pd.read_csv("../input/petfinder-pawpularity-score/test.csv")[['Id']]
submission["Pawpularity"] = mean
submission.to_csv('submission.csv', index=False)
```
