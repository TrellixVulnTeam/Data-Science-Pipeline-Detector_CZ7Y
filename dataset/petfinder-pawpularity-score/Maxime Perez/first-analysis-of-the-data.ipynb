# Analysis of pets data

## Analysis of the categorical variables

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("../input/petfinder-pawpularity-score/train.csv")
df.head()
```

```python
round(df.drop(columns=['Id']).corr()['Pawpularity'], 3)
```

As the correlation matrix show, it seems that the variables seems to not play a big part in order to determine if the pets are popular or not.

## Analysis of the images

```python
plt.hist(df["Pawpularity"], bins=100,color="lightblue")
plt.show()
```

The popularity of the pets do not follow exactly the normal law. In fact we can see 3 extremums.

There is firstly a first peak around 1, meaning that there are photos who are considered not popular at all. 

Then there is a second extremum who assemble who reunite a lot of datas around 25 meaning that the majority of the photos are considered not very popular.

Finally we can see a third peak at 100 meaning that there are a lot of pets considered very popular.

```python
not_popular_pets = df[df["Pawpularity"] <= 2][["Id", "Pawpularity"]].sample(8)
little_popular_pets = df[(df["Pawpularity"] >= 30) & (df["Pawpularity"] <= 35)][["Id", "Pawpularity"]].sample(8)
popular_pets = df[(df["Pawpularity"] >= 70) & (df["Pawpularity"] <= 80)][["Id", "Pawpularity"]].sample(8)
very_popular_pets = df[df["Pawpularity"] == 100][["Id", "Pawpularity"]].sample(8)
img_to_show = pd.concat([not_popular_pets,little_popular_pets,popular_pets,very_popular_pets],ignore_index=True)
```

```python
plt.figure(figsize=(32, 320))
for idx, img in img_to_show.iterrows():
    path = f'../input/petfinder-pawpularity-score/train/{img.Id}.jpg'
    im = plt.imread(path)
    plt.subplot(25, 4, idx+1)
    plt.imshow(im)
    plt.title(f'Pawpularity: {int(img.Pawpularity)}')
    plt.axis('off')
```

Some of the pets who have a terrible popularity are quite beautiful and i find other pets whith a high popularity not that cute.

So from a human perspective it's hard to determine by just looking at the picture if the pet is going to be popular or not.
