This notebook is a ML approach using only the matrix given to us without the images.

```python
import pandas as pd
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
testing_data = pd.read_csv("../input/petfinder-pawpularity-score/test.csv")
testing_data
```

```python
training_data = pd.read_csv("../input/petfinder-pawpularity-score/train.csv")
training_data
```

```python
X = training_data.drop(columns={'Id', 'Pawpularity'})
y = training_data['Pawpularity']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

X_pred = testing_data.drop(columns={'Id'})
```

```python
from xgboost import XGBRegressor

my_model = XGBRegressor(n_estimators=50000, eval_metric=mean_absolute_error)
my_model.fit(X_train, y_train, 
             early_stopping_rounds=10, 
             eval_set=[(X_test, y_test)],
             verbose=False)

y_pred = my_model.predict(X_pred)
```

```python
testing_data['Pawpularity'] = pd.Series(y_pred).astype(float)
testing_data[['Id', 'Pawpularity']].to_csv('submission.csv', index=False)
```
