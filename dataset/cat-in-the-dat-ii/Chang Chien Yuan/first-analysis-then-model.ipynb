```python
categorical_variables=['bin_3','bin_4' ,'nom_0','nom_1','nom_2','nom_3' ,'nom_4','nom_5','nom_6','nom_7',
                       'nom_8','nom_9','ord_1','ord_2','ord_3','ord_4' ,'ord_5']

# new_train = pd.get_dummies(data=train, columns=categorical_variables)
continous_variables =['bin_0', 'bin_1', 'bin_2','ord_0','day','month']
cols = [col for col in train.columns if col!='target']
bin_cols = ['bin_0','bin_1','bin_2','bin_3','bin_4']
ord_cols = ['ord_0','ord_1','ord_2','ord_3','ord_4','ord_5']
nom_cols = ['nom_0','nom_1','nom_2','nom_3','nom_4','nom_5','nom_6','nom_7','nom_8','nom_9']
```

```python
train.sort_index(inplace=True)
train_y = train['target']; test_id = test['id']
train.drop(['target', 'id'], axis=1, inplace=True)
test.drop('id', axis=1, inplace=True)
```

```python
categorical_features = train.columns.tolist();  smoothing=0.20
final_df = pd.DataFrame([])
```

```python
for tr_idx, f_idx in StratifiedKFold(n_splits=5, random_state=2600, shuffle=True).split(train, train_y):
    ce_target_encoder = ce.TargetEncoder(cols = categorical_features, smoothing=smoothing)
    ce_target_encoder.fit(train.iloc[tr_idx, :], train_y.iloc[tr_idx])
    final_df = final_df.append(ce_target_encoder.transform(train.iloc[f_idx, :]), ignore_index=False)
ce_target_encoder = ce.TargetEncoder(cols = categorical_features, smoothing=smoothing)
ce_target_encoder.fit(train, train_y)
train = final_df.sort_index()
test = ce_target_encoder.transform(test)
glm =LogisticRegression( random_state=2, solver='lbfgs', max_iter=20600, fit_intercept=True, penalty='l2', verbose=0)
glm.fit(train, train_y)
```

```python
pd.DataFrame({'id': test_id, 'target': glm.predict_proba(test)[:,1]}).to_csv('submission.csv', index=False)
```
