```python
X_train = pd.read_csv("/kaggle/input/cat-in-the-dat-ii/train.csv")
X_test = pd.read_csv("/kaggle/input/cat-in-the-dat-ii/test.csv")
```

```python
y = X_train['target']
```

```python
 print(X_train['target'].value_counts())
```

```python
X_train.drop(['id','target'], axis=1,inplace=True)
```

```python
test_id = X_test['id']
```

```python
X_test.drop(['id'], axis=1,inplace=True)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X = pd.concat([X_train, X_test], axis=0, sort=False)
```

```python
X.shape
```

```python
X.columns
```

```python
X.dtypes
```

```python
X.isna().sum().sum()
```

```python
X.isna().sum (axis = 0)
```

```python
for column in X.columns:
    X[column].fillna(X[column].mode()[0], inplace=True)
```

```python
X.isna().sum().sum()
```

```python
X.head(10)
```

```python
for col in ['bin_3', 'bin_4', 'nom_0', 'nom_1', 'nom_2', 'nom_3', 'nom_4', 'nom_5', 'nom_6', 'nom_7', 'nom_8', 'nom_9', 
            'ord_1', 'ord_2', 'ord_3', 'ord_4', 'ord_5']:
    print(col)
    print(X[col].value_counts())
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
ordinal_cols = ['bin_3','bin_4','ord_1','ord_2','ord_3','ord_4','ord_5','nom_5','nom_6','nom_7','nom_8','nom_9']
for col in ordinal_cols:
    X[col] = le.fit_transform(X[col])
```

```python
X.head(10)
```

```python
X = pd.concat([X, pd.get_dummies(X['nom_0'], prefix='nom_0')], axis=1)
X = pd.concat([X, pd.get_dummies(X['nom_1'], prefix='nom_1')], axis=1)
X = pd.concat([X, pd.get_dummies(X['nom_2'], prefix='nom_2')], axis=1)
X = pd.concat([X, pd.get_dummies(X['nom_3'], prefix='nom_3')], axis=1)
X = pd.concat([X, pd.get_dummies(X['nom_4'], prefix='nom_4')], axis=1)
X.head()
```

```python
X.drop(['nom_0','nom_1','nom_2','nom_3','nom_4'], axis=1,inplace=True)
X.head()
```

```python
X.shape
```

```python
X.head()
```

```python
X_train,X_test = X[:600000],X[600000:]
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(solver='liblinear')
clf.fit(X_train,y);
print('Score on train dataset:')
clf.score(X_train,y)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix (y,clf.predict(X_train))
```

```python
from sklearn.metrics import roc_curve
y_pred = clf.predict_proba(X_train)[:,1]
fpr, tpr, threshold = roc_curve (y, y_pred)
plt.plot(fpr, tpr)
plt.title('ROC curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.show()
```

```python
y_test = clf.predict_proba(X_test)[:,1]
```

```python
output = pd.DataFrame({"id": test_id, "target": y_test})
output.to_csv("output.csv", index=False)
print("Your submission was successfully saved!")
output.head(30)
```
