### Question 1: Create a DataFrame that has the orders and the products purchased

```python
#Write your answer here
#Create a DataFrame that has the orders and the products purchased 
prd = pd.merge(orders, order_products_prior, on='order_id', how='inner')
prd.head()
```

### Question 2: Get the average days since a prior order for each user.


```python
#Write your answer here
#Getting the average days since a prior order for each user.
avg_days=prd.groupby('user_id')[['days_since_prior_order']].mean()
avg_days.columns=['days_order_mean']
avg_days.head()

#Creating a histrogram for **'days_order_mean'** of **avg_days**.
plt.hist(avg_days.days_order_mean,bins=100)
plt.show()
```

### Question 3: How often does a customer reorder products (reorder ratio of a customer)

```python
#Write your answer here
#Getting the total purchased products of each customer. 
reorders=prd.groupby('user_id')[['product_id']].count()
reorders.columns=['total_bought']

#For each user, getting the distinct number of its purchased products
reorders['total_unique_bought']= prd[prd.reordered==0].groupby('user_id')[['product_id']].count()
reorders.head()

# Creating a ratio of **total_unique_bought**/**total_bought**.
reorders['reorder_ratio']=reorders['total_unique_bought']/reorders['total_bought']
reorders.head()

# Creating a histrogram for **'reorder_ratio'** of **reorders**.
plt.hist(reorders.reorder_ratio,bins=100)
plt.show()

#Sorting the rows of  **reorders**  in order to get first the customers with the lowest **'reorder_ratio'**. 
reorders=reorders.sort_values('reorder_ratio',ascending=True)
reorders.head()

#Saving the index of the customer with the lowest ratio
low_ratio_customer=reorders.index[0]

# Use the **low_ratio_customer** to get from **prd** all of its orders and products purchased
user_low=prd[prd.user_id==low_ratio_customer][['order_id','product_id']]
user_low.head()

#Joinning of **user_low** with **products** so you can get the name of products.
products.head()
user_low=pd.merge(user_low,products,on='product_id',how='left')
user_low.head()

# Getting the name of the products that the customer with lowest ratio has bought.
user_low.product_name.unique()

```
