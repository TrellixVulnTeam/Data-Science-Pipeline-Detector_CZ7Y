```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib notebook
```

# Read Data

```python
o_prior = pd.read_csv('order_products__prior.csv')
o_val = pd.read_csv('order_products__train.csv')
o_order = pd.read_csv('orders.csv')
o_products = pd.read_csv('products.csv')
o_aisles = pd.read_csv('aisles.csv')
o_departments = pd.read_csv('departments.csv')
```

```python
products = np.sort(o_products['product_id'].unique())
users = np.sort(o_order['user_id'].unique())
```

```python
val = o_val.merge(o_order, how = 'left', on = 'order_id' )
val = val.merge(o_products, how='left',on='product_id')
val = val.merge(o_departments, how='left',on='department_id')
val = val.merge(o_aisles, how='left',on='aisle_id')
```

```python
prior = o_prior.merge(o_order, how='left', on='order_id')
prior = prior.merge(o_products, how='left',on='product_id')
prior = prior.merge(o_departments, how='left',on='department_id')
prior = prior.merge(o_aisles, how='left',on='aisle_id')
```

# Plot Data

```python
plt.figure(figsize=(8, 8))
sns.heatmap(prior.drop(['order_id','add_to_cart_order', 'user_id', 'eval_set', 'order_number'], axis=1).corr(), square=True) #heat map
plt.xticks(rotation = 45, fontsize = 5)
plt.yticks(rotation = 0, fontsize = 5)
```

# reordered products

```python
reorder = prior[prior['reordered']==1]
x = reorder['days_since_prior_order']
```

```python
reordered_products = reorder.groupby("product_name")["reordered"].aggregate('sum').reset_index().sort_values(by = ['reordered'], ascending = False).iloc[:300,:].reset_index()
reordered_products.head()
```

```python
plt.figure()
plt.bar(reordered_products.index, reordered_products['reordered'])
```

# reordered by department

```python
reordered_departments = reorder.groupby("department")["reordered"].aggregate('sum').reset_index().sort_values(by = ['reordered'], ascending = False).reset_index()
plt.figure()
sns.barplot(reordered_departments.department, reordered_departments.reordered)
plt.xticks(rotation = 45, fontsize = 8)
```

```python
reordered_aisles = reorder.groupby('aisle')['reordered'].aggregate('sum').reset_index().sort_values(by = ['reordered'], ascending = False).reset_index()
plt.figure(figsize = (8, 8))
sns.barplot(reordered_aisles.aisle, reordered_aisles.reordered)
plt.xticks(rotation = 90, fontsize = 5)
```
