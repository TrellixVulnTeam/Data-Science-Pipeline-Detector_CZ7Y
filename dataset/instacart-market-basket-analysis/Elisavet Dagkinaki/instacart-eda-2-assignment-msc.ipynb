# Insight I:
## Which aisle has the most products?
To answer this question you have to:
- load the products.csv file
- perform the required calculations to find the number of products for each aisle
- visualize your results by including the 10 top aisles with most products.

```python
#Write your answer here

products = pd.read_csv('../input/products.csv')
size =products.groupby('aisle_id')['product_id'].count()
size = size.to_frame('product_size')

size.head(100)

```

```python
#You may create many code blocks to complete your answer


size_results = size.groupby('aisle_id')['product_size'].max().to_frame('total_products')
size_results.head(10)


plt.figure(figsize=(15,10))
graph = sns.barplot(size_results.index, size_results.total_products)
graph.set( xticks=list( range(0,size_results.index.max(),10) ), xticklabels=list( range(0,size_results.index.max(),10) ) )
plt.ylabel('Number of aisle', fontsize=15)
plt.xlabel('Number of products', fontsize=15)
plt.show(10)
```
