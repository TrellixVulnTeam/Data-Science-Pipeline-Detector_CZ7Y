I mean sometimes you can just use **pandas data profiler** to start with. This profiler notebook is packed with information so it takes some time to output all results. Wait ~2min and you'll get a massive summary about 28 data files of the competition with metrics such as: row count, number of columns and column types, duplicates and uniqueness, top values and missing stats, various data demographic highlights etc. Below are just two examples of what this report provides after it draws all results.
