```python
## Data understanding
df_2018_Cities_Disclosing_to_CDP = pd.read_csv("../input/cdp-unlocking-climate-solutions/Cities/Cities Disclosing/2018_Cities_Disclosing_to_CDP.csv")
df_2018_Cities_Disclosing_to_CDP.head()
```

```python
barplot_2018_Cities_Disclosing_to_CDP_Population = df_2018_Cities_Disclosing_to_CDP['Population']
print(barplot_2018_Cities_Disclosing_to_CDP_Population)
```

```python
data_description = barplot_2018_Cities_Disclosing_to_CDP_Population.describe()
print(data_description)
```

```python
#plotting bar graph
sns.set_style("whitegrid")
y = df_2018_Cities_Disclosing_to_CDP['Organization']
x = barplot_2018_Cities_Disclosing_to_CDP_Population
sns.barplot(x = x, y = y)
plt.show()
```

```python
#distribution
#xd = pd.Series(x, name=y)
ax = sns.distplot(x)
plt.show()
```

```python
#Distribution plot with a limit

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15,4))
plt.subplots_adjust(wspace=0.5, hspace=0.5)

sns.distplot((barplot_2018_Cities_Disclosing_to_CDP_Population>=0) & (barplot_2018_Cities_Disclosing_to_CDP_Population<=10000), kde=False, ax=axes[0], axlabel='Population')
sns.distplot((barplot_2018_Cities_Disclosing_to_CDP_Population>=0) & (barplot_2018_Cities_Disclosing_to_CDP_Population<=100000), kde=False, ax=axes[0], axlabel='Population')
sns.distplot((barplot_2018_Cities_Disclosing_to_CDP_Population>=0) & (barplot_2018_Cities_Disclosing_to_CDP_Population<=1000000), kde=False, ax=axes[0], axlabel='Population')
```

```python
#Measure Skewness
skewValue = x.skew()
print(skewValue)
```
