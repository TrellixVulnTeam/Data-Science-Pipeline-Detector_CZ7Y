Good news, it appears that higher percentage of respondents have a climate change plan in year-over-year comparisons.

```python
def f(year):
    df2 = dataconcatQ2
    fig = px.pie(df2[df2['Year Reported to CDP'] ==year], values = 'count', names ='Response Answer')
    fig.show()
interact(f, year=[2020,2019,2018])
```
