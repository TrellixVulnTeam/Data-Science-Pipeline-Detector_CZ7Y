```python
countriesDataset = pd.read_csv('../input/country-codes/wikipedia-iso-country-codes.csv')
countriesDataset = countriesDataset[['Country','Code']]


countriesDataset['Country'] = countriesDataset['Country'].replace({"Korea, Democratic People's Republic of": 'Republic of Korea'})
countriesDataset['Country'] = countriesDataset['Country'].replace({'United Kingdom': 'United Kingdom of Great Britain and Northern Ireland'})
countriesDataset['Country'] = countriesDataset['Country'].replace({'Taiwan':'Taiwan, Greater China'})
countriesDataset['Country'] = countriesDataset['Country'].replace({'Russia':'Russian Federation'})
countriesDataset['Country'] = countriesDataset['Country'].replace({'Palestinian Territory, Occupied':'State of Palestine'})
countriesDataset['Country'] = countriesDataset['Country'].replace({'Congo, the Democratic Republic of the':'Democratic Republic of the Congo'})
countriesDataset['Country'] = countriesDataset['Country'].replace({'Bolivia':'Bolivia (Plurinational State of)'})
countriesDataset['Country'] = countriesDataset['Country'].replace({'Vietnam':'Viet Nam'})
countriesDataset['Country'] = countriesDataset['Country'].replace({'Hong Kong':'Hong Kong Special Administrative Region'})
countriesDataset['Country'] = countriesDataset['Country'].replace({'Venezuela':'Venezuela (Bolivarian Republic of)'})
countriesDataset['Country'] = countriesDataset['Country'].replace({'United States Of America':'United States of America'})

```

# Questions based on CDP's recommendations

### Question 0.1(INTRODUCTION) - Please give a general desription and introduction to your city including your city's reporting boundary in the table below. [administrative boundary, description of the city]

```python
question_01 = citiesDataset2019.loc[citiesDataset2019['Question Number']== '0.1']

LGA = question_01.loc[question_01['Response Answer']== 'Local government area within a city / metropolitan area']
city_Municipality = question_01.loc[question_01['Response Answer']== 'City / Municipality']
independent_City  = question_01.loc[question_01['Response Answer']== 'Independent city']
special_City   = question_01.loc[question_01['Response Answer']== 'Special city']
federal_District    = question_01.loc[question_01['Response Answer']== 'Federal district']
sovereign_City_State     = question_01.loc[question_01['Response Answer']== 'Sovereign city-state']
metropolitan_Area      = question_01.loc[question_01['Response Answer']== 'Metropolitan area']
province_County       = question_01.loc[question_01['Response Answer']== 'Province / County']
independent_Province        = question_01.loc[question_01['Response Answer']== 'Independent province'] 
intercommunality        = question_01.loc[question_01['Response Answer']== 'Intercommunality']
sub_Municipal_District        = question_01.loc[question_01['Response Answer']== 'Sub-municipal district']
other = question_01.loc[question_01['Response Answer']== 'Other']

city_Description = pd.concat([LGA,city_Municipality,independent_City,special_City,federal_District,sovereign_City_State,metropolitan_Area,province_County,
                             independent_Province,intercommunality,sub_Municipal_District,other])
city_Description1 = city_Description.groupby(['Response Answer'],as_index = False).count()
```

```python
plt.figure(figsize=(20,18))
city_Descriptionfig1 = px.choropleth(city_Description,  locations= "Code", color = 'Response Answer', hover_name= "City", scope='world',width=900,height=800, color_continuous_scale= px.colors.sequential.Viridis)
city_Descriptionfig1.show()
```

### Note: The very light blue areas on the map are either not included in the dataset or those cities did not provide an answer to question 

```python
plt.figure(figsize=(10,8))
plt.xticks(rotation='vertical')
city_Descriptionfig2 = sns.lineplot(x = 'Response Answer', y= 'Question Number', data = city_Description1)
```

# USA - SOCIAL VULNERABILITY INDEX (SVI)

```python
socialVulnerabilityData = pd.read_csv('/kaggle/input/cdp-unlocking-climate-solutions/Supplementary Data/CDC Social Vulnerability Index 2018/SVI2018_US_COUNTY.csv')
socialVulnerabilityData.groupby('STATE')['STATE'].count()
```

```python
socialVulnerabilityDataSum = socialVulnerabilityData.groupby(['STATE'],as_index=False).sum()
```

### E_POV - Persons below poverty estimate

```python
socialVulnerabilitySTATE_EPOV= socialVulnerabilityDataSum[['STATE','E_POV']]
socialVulnerabilitySTATE_EPOV.sort_values(by = 'E_POV', ascending=False)
```

```python
#socialVulnerabilitySTATE_EPOV.describe()

plt.figure(figsize=(12,10))
socialVulnerabilityDataSumfig1 = sns.lineplot(x = "STATE", y = "E_POV", data = socialVulnerabilityDataSum)
plt.xticks(rotation='vertical')
socialVulnerabilityDataSumfig1.set(yscale='log')
```

#### Notes:
1. The US state that has the highest number of persons below poverty estimate between year 2014 and 2018 is CALIFORNIA. The total number of these people is 5,487,141
2. The US state that has the lowest number of persons below poverty estimate between year 2014 and 2018 is WYOMING. The total number of these people is 63,311
3. The average number of people living below poverty estimate per state in the United State between 2014 and 2018 is 867,566
4. CALIFORNIA state has 58 counties. It is good to check throughout the counties as well

```python
socialVulnerabilityDataCALIFORNIA = socialVulnerabilityData.loc[socialVulnerabilityData['STATE']=='CALIFORNIA']
socialVulnerabilityDataCALIFORNIA_EPOV = socialVulnerabilityDataCALIFORNIA[['STATE','COUNTY','E_POV']]
socialVulnerabilityDataCALIFORNIA_EPOV.sort_values(by = 'E_POV', ascending = False)
```

```python
socialVulnerabilityDataCALIFORNIA_EPOV.describe()

LA_EPOV = socialVulnerabilityDataCALIFORNIA_EPOV.loc[socialVulnerabilityDataCALIFORNIA_EPOV['COUNTY']== 'Los Angeles']
LA_EPOV.drop(columns=['STATE'])
column_names = {'COUNTY':'county_name','E_POV':'population'}
LA_EPOV = LA_EPOV.rename(columns = column_names)

plt.figure(figsize=(12,10))
socialVulnerabilityDataCALIFORNIA_EPOVfig1 = sns.barplot(x = "COUNTY", y = "E_POV", data = socialVulnerabilityDataCALIFORNIA_EPOV)
plt.xticks(rotation ='vertical')
```

### Note:

### E_NOHSDP - Pearsons (age 25+) with no high school diploma estimate

```python
socialVulnerabilitySTATE_ENOHSDP= socialVulnerabilityDataSum[['STATE','E_NOHSDP']]
e_NOHSDP= socialVulnerabilitySTATE_ENOHSDP.sort_values(by = 'E_NOHSDP', ascending=False)

socialVulnerabilitySTATE_EMINRTY= socialVulnerabilityDataSum[['STATE','E_MINRTY']]
e_MINRTY= socialVulnerabilitySTATE_EMINRTY.sort_values(by = 'E_MINRTY', ascending=False)

socialVulnerabilitySTATE_ECROWD= socialVulnerabilityDataSum[['STATE','E_CROWD']]
e_CROWD= socialVulnerabilitySTATE_ECROWD.sort_values(by = 'E_CROWD', ascending=False)

socialVulnerabilitySTATE_ENOVEH= socialVulnerabilityDataSum[['STATE','E_NOVEH']]
e_NOVEH= socialVulnerabilitySTATE_ENOVEH.sort_values(by = 'E_NOVEH', ascending=False)

plt.rcParams["figure.figsize"] = (20,20)

e_NOHSDP_MINRTY_CROWD_NOVEH = pd.concat({'25+ years with no high school diploma': e_NOHSDP.set_index('STATE'),
                                         'Minority': e_MINRTY.set_index('STATE'), 'Household, more people than room estimate': e_CROWD.set_index('STATE'),
                                        'Household with no vehicle': e_NOVEH.set_index('STATE')}, axis=1).plot.bar(width=2,color=['b','g','r','y'])
e_NOHSDP_MINRTY_CROWD_NOVEH.set(title = 'E_NOHSDP, E_MINRTY, E_CROWD, E_NOVEH indices by State')
e_NOHSDP_MINRTY_CROWD_NOVEH.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
e_NOHSDP_MINRTY_CROWD_NOVEH.set(yscale='log')

plt.show()
```

### Note:
1. CALIFORNIA state has the highest population, the largest number of *people living below poverty estimate, *civillian 16+ and unemployed, *25+ with no high school diploma, *minority population and *people occupying more than room estimate

2. The number of Minority group for all state is higher than the other indices.
