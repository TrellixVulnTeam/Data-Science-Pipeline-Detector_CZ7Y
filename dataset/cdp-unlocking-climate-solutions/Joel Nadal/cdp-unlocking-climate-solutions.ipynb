### Cliamate change risks detected by cities

```python
df_cities_2_1_2020=df_2020_Full_Cities_Dataset.loc[df_2020_Full_Cities_Dataset['Question Number'] == '2.1'].copy()
df_cities_2_1_2020.sort_values(by=['Organization','Column Number','Row Number'],inplace=True)
df_cities_2_1_2020_Climate_Hazards=df_cities_2_1_2020.loc[df_cities_2_1_2020['Column Name'] == 'Climate Hazards'].copy()
df_cities_2_1_2020_Climate_Hazards=df_cities_2_1_2020_Climate_Hazards.groupby('Response Answer').count()
df_cities_2_1_2020_Climate_Hazards.sort_values(by='Organization',ascending=False,inplace=True)
df_cities_2_1_2020=df_2020_Full_Cities_Dataset.loc[df_2020_Full_Cities_Dataset['Question Number'] == '2.1'].copy()
df_cities_2_1_2020.sort_values(by=['Organization','Row Number'],inplace=True)

city_name=[]
city_risk_list=[]
city_current_list=[]
city_future_list=[]
city_frequency_list=[]

for i in range(len(df_cities_2_1_2020['Organization'].unique())):
    city=df_cities_2_1_2020.loc[df_cities_2_1_2020['Organization'] == df_cities_2_1_2020['Organization'].unique()[i]].copy()
    for j in range(len(city['Row Number'].unique())):
        city_risk=city.loc[city['Row Number'] == city['Row Number'].unique()[j]].copy()
        city_risk_magnitude_current=city_risk.loc[city_risk['Column Number'] == 3].copy()
        city_risk_magnitude_future=city_risk.loc[city_risk['Column Number'] == 10].copy()
        if city_risk_magnitude_current['Response Answer'].isin(['High','Medium High','Medium']).bool() | city_risk_magnitude_future['Response Answer'].isin(['High','Medium High','Medium']).bool():
            city_name.append(city_risk_magnitude_current['Organization'].copy().to_string(index=False))
            city_risk_list.append(city_risk.loc[city_risk['Column Number'] == 1].copy()['Response Answer'].to_string(index=False))
            city_current_list.append(city_risk.loc[city_risk['Column Number'] == 4].copy()['Response Answer'].to_string(index=False))
            city_future_list.append(city_risk.loc[city_risk['Column Number'] == 10].copy()['Response Answer'].to_string(index=False))
            city_frequency_list.append(city_risk.loc[city_risk['Column Number'] == 8].copy()['Response Answer'].to_string(index=False))
            del (city_risk,city_risk_magnitude_current,city_risk_magnitude_future)
    del(city)

risks_cities = pd.DataFrame(list(zip(city_name, city_risk_list,city_current_list,city_future_list,city_frequency_list)), 
               columns =['Organization','Risk','Current risk magnitude','Future risk magnitude','Future risk frequency change']) 
risks_cities=risks_cities.groupby('Risk').count()
risks_cities.sort_values(by='Organization',ascending=True,inplace=True)
risk_cities=risks_cities.copy()
risk_cities=risk_cities.iloc[math.floor(-len(risks_cities)*0.75):,:]

labels=risk_cities.index
x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars
fig, ax = plt.subplots(figsize=(7, 15))
rects1 = ax.barh(x + width/2, risk_cities['Organization'], width)
ax.set_title('Climate hazards reported by cities frequency in 2020 \n Only taken into account the hazards reported to be from medium to high \n and the 75% most reported',fontsize=20)
plt.yticks(x, labels,fontsize=20)
ax.set_xlabel('Total amount of the cities that reported the climate hazard',fontsize=20)
plt.xticks(fontsize=20)
plt.show()
```

It is interesting that both extrem precipitation and water scarcity are in the TOP 3 of the risks detected by the cities.
