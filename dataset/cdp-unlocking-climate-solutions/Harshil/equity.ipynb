# Socioeconomic KPIs

| KPI Name               | KPI Unit    | KPI Interpretation                                              | Public Data Source                 |
|------------------------|-------------|-----------------------------------------------------------------|------------------------------------|
| socio_economic         | unitless    | CDC's Social Vulnerability Index for socioeconomic indicators   | CDC                                |

| job_density            | jobs/sqmile | Job density                                                     | Chetty et al. (2014) / Census      |
| avg_job_growth         | %           | Average job growth                                              | Chetty et al. (2014) / Census      |

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
from sklearn.preprocessing import RobustScaler, QuantileTransformer, StandardScaler, MinMaxScaler
import ast
plt.style.use("/kaggle/input/mpl-style-file/harsh.mpl")
from subprocess import check_output
%matplotlib inline
```

```python
path = "../input/externald/Supplementary Data/" 
equity_cities = pd.read_csv("../input/externald/Supplementary Data/cities_updated_geo_us_2020.csv") # takes the output from reduce_and_add_cities

counties = equity_cities['counties'] # counties
counties_weights = (equity_cities['counties_weights']) # county "weights" for each city
states = equity_cities['state'] # state that this city belongs to
flag_multi_counties = equity_cities['flag_multiple_counties'] # flag indicating whether this city
match_fips = equity_cities['FIPS']
match_fips = [ast.literal_eval(match_fips[i]) if flag_multi_counties[i] else int(match_fips[i]) for i in range(len(match_fips))]
```

```python
def standardize_rank(arr, direction=1):
    rank = (stats.mstats.rankdata(np.ma.masked_invalid(arr)))
    rank[rank == 0] = np.nan
    if(direction == 1):
        return rank/(np.nanmax(rank)) 
    else:
        return 1-(rank/(np.nanmax(rank)))
```

```python
df = pd.read_csv(path + "CDC Social Vulnerability Index 2018/SVI2018_US_COUNTY.csv")
```

```python
cdc_svi_metrics = []
for i in range(len(counties)):
    if(not flag_multi_counties[i]):
        fips = int(match_fips[i])
        cdf = df[df["FIPS"] == fips]
        cdc_svi_metrics.append(np.array([float(cdf['RPL_THEME1']), float(cdf['RPL_THEME2']), 
                                 float(cdf['RPL_THEME3']), float(cdf['RPL_THEME4'])]))
    else:
        fips = (match_fips[i])
        temp = []
        for j in range(len(fips)):
            cdf = df[df["FIPS"] == fips[j]]
            temp.append(np.array([float(cdf['RPL_THEME1']), float(cdf['RPL_THEME2']), 
                                 float(cdf['RPL_THEME3']), float(cdf['RPL_THEME4'])]))
        all_county_metrics = np.array(temp)
        weighted_metrics = np.matmul(np.array(ast.literal_eval(counties_weights[i])), all_county_metrics)
        cdc_svi_metrics.append(weighted_metrics)
```

```python
cdc_svi_metrics = np.array(cdc_svi_metrics)
direction_cdc_svi_metrics = 1.0

socio_economic_score = standardize_rank(cdc_svi_metrics[:,0], direction=direction_cdc_svi_metrics)
house_composition_disability_score = standardize_rank(cdc_svi_metrics[:,1], direction=direction_cdc_svi_metrics)
minority_language_score = standardize_rank(cdc_svi_metrics[:,2], direction=direction_cdc_svi_metrics)
house_type_transportation_score = standardize_rank(cdc_svi_metrics[:,3], direction=direction_cdc_svi_metrics)
```

```python
dff = pd.read_csv(path + "mobility_chetty.csv")

rm = []
for i in range(len(match_fips)):
    if(not flag_multi_counties[i]):
        fips = int(match_fips[i])
        cdf = dff[dff["County FIPS Code"] == fips]
        rm.append(np.float(cdf["Rank-Rank Slope"]))
    else:
        fips = match_fips[i]
        temp = []
        for j in range(len(fips)):
            cdf = dff[dff["County FIPS Code"] == fips[j]]
            temp.append(np.float(cdf["Rank-Rank Slope"]))
        ar = np.array(temp)
        weighted_avg = np.average(ar, weights=np.array(ast.literal_eval(counties_weights[i])))
        rm.append(weighted_avg)
        
am = []
for i in range(len(match_fips)):
    if(not flag_multi_counties[i]):
        fips = int(match_fips[i])
        cdf = dff[dff["County FIPS Code"] == fips]
        am.append(np.float(cdf["Absolute Upward Mobility"]))
    else:
        fips = (match_fips[i])
        temp = []
        for j in range(len(fips)):
            cdf = dff[dff["County FIPS Code"] == fips[j]]
            temp.append(np.float(cdf["Absolute Upward Mobility"]))
        ar = np.array(temp)
        weighted_avg = np.average(ar, weights=np.array(ast.literal_eval(counties_weights[i])))
        am.append(weighted_avg)
```

```python
relative_mobility = np.array(rm)
direction_relative_mobility = 1.0

absolute_mobility = np.array(am)
direction_absolute_mobility = -1.0

relative_mobility_score = standardize_rank(relative_mobility, direction=direction_relative_mobility)
absolute_mobility_score = standardize_rank(absolute_mobility, direction=direction_absolute_mobility)
```

```python
dfff = pd.read_csv(path + "cty_covariates.csv")
ct = np.char.zfill(np.array(dfff['county']).astype(str), 3)
st = np.char.zfill(np.array(dfff['state']).astype(str), 2)
dfff['fips'] = np.core.defchararray.add(st, ct).astype(int)

avg_job_growth = []
for i in range(len(match_fips)):
    if(not flag_multi_counties[i]):
        fips = int(match_fips[i])
        cdf = dfff[dfff["fips"] == fips]
        avg_job_growth.append(np.float(cdf["ann_avg_job_growth_2004_2013"]))
    else:
        fips = (match_fips[i])
        temp = []
        for j in range(len(fips)):
            cdf = dfff[dfff["fips"] == fips[j]]
            temp.append(np.float(cdf["ann_avg_job_growth_2004_2013"]))
        ar = np.array(temp)
        weighted_avg = np.average(ar, weights=np.array(ast.literal_eval(counties_weights[i])))
        avg_job_growth.append(weighted_avg)
        
job_dens = []
for i in range(len(match_fips)):
    if(not flag_multi_counties[i]):
        fips = int(match_fips[i])
        cdf = dfff[dfff["fips"] == fips]
        job_dens.append(np.float(cdf["job_density_2013"]))
    else:
        fips = (match_fips[i])
        temp = []
        for j in range(len(fips)):
            cdf = dfff[dfff["fips"] == fips[j]]
            temp.append(np.float(cdf["job_density_2013"]))
        ar = np.array(temp)
        weighted_avg = np.average(ar, weights=np.array(ast.literal_eval(counties_weights[i])))
        job_dens.append(weighted_avg)
        
```

```python
avg_job_growth = np.array(avg_job_growth)
direction_avg_job_growth = -1.0

job_dens = np.array(job_dens)
direction_job_dens = -1.0

avg_job_growth_score = standardize_rank(avg_job_growth, direction=direction_avg_job_growth)
job_dens_score = standardize_rank(job_dens, direction=direction_job_dens)
```

```python
tf = pd.read_excel(path + "Feeding America/Map the Meal Gap data/MMG2020_2018Data_ToShare.xlsx", sheet_name="2018 County", skiprows=1)

food_insec = []
for i in range(len(match_fips)):
    if(not flag_multi_counties[i]):
        fips = int(match_fips[i])
        cdf = tf[tf["FIPS"] == fips]
        food_insec.append(np.float(cdf["2018 Food Insecurity Rate"]))
    else:
        fips = (match_fips[i])
        temp = []
        for j in range(len(fips)):
            cdf = tf[tf["FIPS"] == fips[j]]
            temp.append(np.float(cdf["2018 Food Insecurity Rate"]))
        ar = np.array(temp)
        weighted_avg = np.average(ar, weights=np.array(ast.literal_eval(counties_weights[i])))
        food_insec.append(weighted_avg)
        
child_food_insec = []
for i in range(len(match_fips)):
    if(not flag_multi_counties[i]):
        fips = int(match_fips[i])
        cdf = tf[tf["FIPS"] == fips]
        child_food_insec.append(np.float(cdf["2018 Child food insecurity rate"]))
    else:
        fips = (match_fips[i])
        temp = []
        for j in range(len(fips)):
            cdf = tf[tf["FIPS"] == fips[j]]
            temp.append(np.float(cdf["2018 Child food insecurity rate"]))
        ar = np.array(temp)
        weighted_avg = np.average(ar, weights=np.array(ast.literal_eval(counties_weights[i])))
        child_food_insec.append(weighted_avg)
```

```python
food_insec = np.array(food_insec)
direction_food_insec = 1.0

child_food_insec = np.array(child_food_insec)
direction_child_food_insec = 1.0

food_insec_score = standardize_rank(food_insec, direction=direction_food_insec)
child_food_insec_score = standardize_rank(child_food_insec, direction=direction_child_food_insec)
```

```python
ch = pd.read_csv(path + "CountyHealth_analytic.csv")

def extract_from_county_health_survey(field):
    metric = []
    for i in range(len(match_fips)):
        if(not flag_multi_counties[i]):
            fips = int(match_fips[i])
            cdf = ch[ch["FIPS"] == fips]
            metric.append(np.float(cdf[field]))
        else:
            fips = (match_fips[i])
            temp = []
            for j in range(len(fips)):
                cdf = ch[ch["FIPS"] == fips[j]]
                temp.append(np.float(cdf[field]))
            ar = np.array(temp)
            weighted_avg = np.average(ar, weights=np.array(ast.literal_eval(counties_weights[i])))
            metric.append(weighted_avg)
    return np.array(metric)
```

```python
premature_death_rate = extract_from_county_health_survey("Premature death raw value")
direction_premature_death_rate = 1.0
premature_death_rate_score = standardize_rank(premature_death_rate, direction=direction_premature_death_rate)

health_level = extract_from_county_health_survey("Poor or fair health raw value")
direction_health_level = 1.0
health_level_score = standardize_rank(health_level, direction=direction_health_level)

healthy_food_access = extract_from_county_health_survey("Food environment index raw value")
direction_healthy_food_access = -1.0
healthy_food_access_score = standardize_rank(healthy_food_access, direction=direction_healthy_food_access)

exercise_access = extract_from_county_health_survey("Access to exercise opportunities raw value")
direction_exercise_access = -1.0
exercise_access_score = standardize_rank(exercise_access, direction=direction_exercise_access)

percent_uninsured = extract_from_county_health_survey("Uninsured raw value")
direction_percent_uninsured = 1.0
percent_uninsured_score = standardize_rank(percent_uninsured, direction=direction_percent_uninsured)

primary_care_physicians = extract_from_county_health_survey("Primary care physicians raw value")
direction_primary_care_physicians = -1.0
primary_care_physicians_score = standardize_rank(primary_care_physicians, direction=direction_primary_care_physicians)

segregation_bl = extract_from_county_health_survey("Residential segregation - Black/White raw value")
direction_segregation_bl = 1.0
segregation_bl_score = standardize_rank(segregation_bl, direction=direction_segregation_bl)

segregation_nw = extract_from_county_health_survey("Residential segregation - non-White/White raw value")
direction_segregation_nw = 1.0
segregation_nw_score = standardize_rank(segregation_nw, direction=direction_segregation_nw)

housing_cost_burden = extract_from_county_health_survey("Severe housing cost burden raw value")
direction_housing_cost_burden = 1.0
housing_cost_burden_score = standardize_rank(housing_cost_burden, direction=direction_housing_cost_burden)

mental_distress = extract_from_county_health_survey("Frequent mental distress raw value")
direction_mental_distress = 1.0
mental_distress_score = standardize_rank(mental_distress, direction=direction_mental_distress)
```

```python
trends = pd.read_csv(path + "CHR_trends_csv_2020.csv", encoding='latin-1')
ct = np.char.zfill(np.array(trends['countycode']).astype(str), 3)
st = np.char.zfill(np.array(trends['statecode']).astype(str), 2)
trends['fips'] = np.core.defchararray.add(st, ct).astype(int)
```

```python
def extract_trends_from_county_health_survey(field, num_measures_for_mean=3):
    metric = []
    for i in range(len(match_fips)):
        if(not flag_multi_counties[i]):
            fips = int(match_fips[i])
            cdf = trends[(trends['measurename'] == field) & (trends['fips'] == fips)]
            
            rv = list(cdf['rawvalue'])
            if(type(rv[0]) == str):
                rv = [float(x.replace(",", "")) for x in rv]
            
            rvd = (np.mean(rv[-num_measures_for_mean:]) - np.mean(rv[0:num_measures_for_mean]))/(np.mean(rv[0:num_measures_for_mean]))
            metric.append(np.float(rvd))
        else:
            fips = (match_fips[i])
            temp = []
            for j in range(len(fips)):
                cdf = trends[(trends['measurename'] == field) & (trends['fips'] == fips[j])]
                rv = list(cdf['rawvalue'])
                
                if(type(rv[0]) == str):
                    rv = [float(x.replace(",", "")) for x in rv]
                
                rvd = (np.mean(rv[-num_measures_for_mean:]) - np.mean(rv[0:num_measures_for_mean]))/(np.mean(rv[0:num_measures_for_mean]))
                temp.append(np.float(rvd))
            ar = np.array(temp)
            weighted_avg = np.average(ar, weights=np.array(ast.literal_eval(counties_weights[i])))
            metric.append(weighted_avg)
    return np.array(metric)
```

```python
premature_death_trend = extract_trends_from_county_health_survey("Premature death")
direction_premature_death_trend = 1.0
premature_death_trend_score = standardize_rank(premature_death_trend, direction=direction_premature_death_trend)

unemployment_trend = extract_trends_from_county_health_survey("Unemployment rate")
direction_unemployment_trend = 1.0
unemployment_trend_score = standardize_rank(unemployment_trend, direction=direction_unemployment_trend)

pcp_trend = extract_trends_from_county_health_survey("Primary care physicians")
direction_pcp_trend = -1.0
pcp_trend_score = standardize_rank(pcp_trend, direction=direction_pcp_trend)

uninsured_trend = extract_trends_from_county_health_survey("Uninsured")
direction_uninsured_trend = 1.0
uninsured_trend_score = standardize_rank(uninsured_trend, direction=direction_uninsured_trend)

children_poverty_trend = extract_trends_from_county_health_survey("Children in poverty")
direction_children_poverty_trend = 1.0
children_poverty_trend_score = standardize_rank(children_poverty_trend, direction=direction_children_poverty_trend)
```

```python
nri = pd.read_csv(path + "NRI_Table_Counties.csv")

def nri_select(field):
    res = []
    for i in range(len(match_fips)):
        if(not flag_multi_counties[i]):
            fips = int(match_fips[i])
            cdf = nri[nri["STCOFIPS"] == fips]
            try:
                res.append(float(cdf[field]))
            except:
                res.append(np.nan)
        else:
            fips = match_fips[i]
            temp = []
            for j in range(len(fips)):
                cdf = nri[nri["STCOFIPS"] == fips[j]]
                try:
                    temp.append(float(cdf[field]))
                except:
                    temp.append(np.nan)
            ar = np.array(temp)
            weighted_avg = np.average(ar, weights=np.array(ast.literal_eval(counties_weights[i])))
            res.append(weighted_avg)
    return np.array(res)
```

```python
sovi_risk_index = nri_select("SOVI_SCORE")
direction_sovi_risk_index = 1.0
sovi_risk_index_score = standardize_rank(sovi_risk_index, direction=direction_sovi_risk_index)
```
