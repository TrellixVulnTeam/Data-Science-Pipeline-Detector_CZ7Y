#### Climate change is the single greatest threat facing humanity today. The future of civilization is largely uncertain and bound to be messy. The central theme that I kept coming back to throughout working on this project was the importance of community-based action. To that end, I chose to focus most of my analysis on cities. In particular, I focused on cities in the US because of the abundance of public, external datasets for American cities. 

For this analysis, I will be largely be focusing on three of the four broad questions that CDP put forth. 


       How can we measure the intersection between environmental risks and social equity, as a contributor to resiliency?
       
       How do you help cities adapt to a rapidly changing climate amidst a global pandemic, but do it in a way that is socially equitable?

       What are the practical and actionable points where city and corporate ambition join, i.e. where do cities have problems that corporations affected by those problems could solve, and vice versa?


The framework I will present, though, can be further refined to focus on the fourth question of what concrete projects can cities and corporations collaborate on? 

Let's start with the first (and perhaps the most important) question: How can we measure the intersection between environmental risks and social equity, as a contributor to resiliency? To do this, I set out to build summary metrics that can capture the degree of environmental **vulnerability**, magnitude of social **inequity**, and **adaptability** for any given city. 

After combining the CDP cities survey with a dozen or so external datasets, I derived more than 100 KPIs for each city under those three broad categories (vulnerability, inequity, adaptability). The gory details of each metric can be found in the linked notebooks near the end of the notebook. I will first present maps of the US colored by aggregated KPIs in the three different categories.

```python
def figure_equity(field, title, cmap='plasma_r'):
    fig = go.Figure()

    string_hover = cities + "<br>" + "Overall: " +  city_aggr_equity["aggr_all_inequity"].astype(str).str[0:4] + "<br>" + "Health: " + \
    city_aggr_equity["aggr_health_kpis"].astype(str).str[0:4] + "<br>" + "Socioeconomic: " + city_aggr_equity["aggr_socio_economic_kpis"].astype(str).str[0:4]\
    + "<br>" + "Housing: " + city_aggr_equity["aggr_housing_kpis"].astype(str).str[0:4]

    fig = go.Figure(go.Scattergeo(
        lat=clat,
        lon=clng,
        hovertext=string_hover,
        marker=dict(opacity=0.8,color=(city_aggr_equity[field]),autocolorscale = False,
                    colorscale=cmap,
                    size=10, colorbar=dict(
                      title=title + "<br>" + "Inequity", x=0.9), line_color='rgb(40,40,40)')
    ))

    fig.update_layout(
            title = {"text":
                     title + ' Inequity Index',
                     'y':0.9,
            'x':0.5,
            'xanchor': 'center',
            'yanchor': 'top'},
            geo = dict(
                scope='usa',
                projection_type='albers usa',
                showland = True,
                landcolor = "rgb(240, 240, 240)",
                subunitcolor = "rgb(217, 217, 217)",
                countrycolor = "rgb(217, 217, 217)",
                countrywidth = 0.5,
                subunitwidth = 0.5
            ),
            font=dict(
            family="Garamond",
            color="black",
            size=22
        )
    )

    fig.show()
    
def figure_vul(field, title, cmap="viridis"):
    fig = go.Figure()

    string_hover = cities + "<br>" + "Overall: " +  city_aggr_vul["aggr_vul_all_kpis"].astype(str).str[0:4] + "<br>" + "Heat: " + \
    city_aggr_vul["aggr_heat_kpis"].astype(str).str[0:4] + "<br>" + "Cold: " + city_aggr_vul["aggr_cold_kpis"].astype(str).str[0:4]\
    + "<br>" + "Drought: " + city_aggr_vul["aggr_drought_kpis"].astype(str).str[0:4] + "<br>" + "Severe Weather: " + city_aggr_vul["aggr_severe_kpis"].astype(str).str[0:4] \
    + "<br>" + "Floods: " + city_aggr_vul["aggr_flood_kpis"].astype(str).str[0:4]


    fig = go.Figure(go.Scattergeo(
        lat=clat,
        lon=clng,
        hovertext=string_hover,
        marker=dict(opacity=0.8,color=(city_aggr_vul[field]),autocolorscale = False,
                    colorscale=cmap,
                    size=10, colorbar=dict(
                      title=title + "<br>" + "Vulnerability", x=0.9), line_color='rgb(40,40,40)')
    ))

    fig.update_layout(
            title = {"text":
                     title + ' Vulnerability Index',
                     'y':0.9,
            'x':0.5,
            'xanchor': 'center',
            'yanchor': 'top'},
            geo = dict(
                scope='usa',
                projection_type='albers usa',
                showland = True,
                landcolor = "rgb(240, 240, 240)",
                subunitcolor = "rgb(217, 217, 217)",
                countrycolor = "rgb(217, 217, 217)",
                countrywidth = 0.5,
                subunitwidth = 0.5
            ),
            font=dict(
            family="Garamond",
            color="black",
            size=24
        )
    )

    fig.show()

def figure_adapt(field, title, cmap="haline_r"):
    fig = go.Figure()

    string_hover = cities + "<br>" + "Overall: " +  city_aggr_adapt["aggr_adaptability_kpis"].astype(str).str[0:4] + "<br>" + "Resilience: " + \
    city_aggr_adapt["aggr_resilience_kpis"].astype(str).str[0:4] + "<br>" + "Sentiment: " + city_aggr_adapt["aggr_sentiment_kpis"].astype(str).str[0:4] + "<br>" 


    fig = go.Figure(go.Scattergeo(
        lat=clat,
        lon=clng,
        hovertext=string_hover,
        marker=dict(opacity=0.8,color=(city_aggr_adapt[field]),autocolorscale = False,
                    colorscale=cmap,
                    size=10, colorbar=dict(
                      title=title + "<br>", x=0.9), line_color='rgb(40,40,40)')
    ))

    fig.update_layout(
            title = {"text":
                     title,
                     'y':0.9,
            'x':0.5,
            'xanchor': 'center',
            'yanchor': 'top'},
            geo = dict(
                scope='usa',
                projection_type='albers usa',
                showland = True,
                landcolor = "rgb(240, 240, 240)",
                subunitcolor = "rgb(217, 217, 217)",
                countrycolor = "rgb(217, 217, 217)",
                countrywidth = 0.5,
                subunitwidth = 0.5
            ),
            font=dict(
            family="Garamond",
            color="black",
            size=24
        )
    )

    fig.show()
```

```python
def scatter(x,y,hovertext=cities_2020["city_bing"],color=None,cmap="viridis", xlabel="", ylabel="", title="plot", cbar_title="title"):
    fig = go.Figure(data=go.Scatter(
        x = x,
        y = y,
        hovertext=hovertext,
        mode='markers',
        marker=dict(
            size=14,
            color=color, #set color equal to a variable
            colorscale=cmap, # one of plotly colorscales
            showscale=True, colorbar=dict(
                      title=cbar_title + "<br>", x=1.05), line_color='rgb(40,40,40)'
        )))

    
    fig.update_layout(
        title = {"text":
                     title,
                     'y':0.9,
            'x':0.5,
            'xanchor': 'center',
            'yanchor': 'top'},
        xaxis=dict(
            title=xlabel,
            gridcolor='white',
            gridwidth=2,
        ),
        
        yaxis=dict(
            title=ylabel,
            gridcolor='white',
            gridwidth=2,
        ),
        paper_bgcolor='rgb(243, 243, 243)',
        plot_bgcolor='rgb(243, 243, 243)',
            font=dict(
            family="Garamond",
            color="black",
            size=22
        )
    )
    
    fig.update_layout(xaxis=dict(scaleanchor='y', constrain='domain'),
                  coloraxis_colorbar_x=0.5, width=900, height=800
                 )


    fig.show()
```

```python
def three_by_two_plot(fields, xlabels, idx_city=0, city="Somerville", color="cornflowerblue", log=False):
    fig, ax = plt.subplots(3,2,figsize=(20,30))

    sns.kdeplot(df[fields[0]][:], label="", color=color, shade=True, ax=ax[0,0])
    ax[0,0].axvline(df[fields[0]][idx_city], linestyle='--', label=city)
    ax[0,0].set_xlabel(xlabels[0])
    ax[0,0].legend(loc='best', fontsize=24, frameon=False)

    sns.kdeplot(df[fields[1]][:], label="", color=color, shade=True, ax=ax[0,1])
    ax[0,1].axvline(df[fields[1]][idx_city], linestyle='--')
    ax[0,1].set_xlabel(xlabels[1])
    ax[0,1].legend(loc='best', fontsize=24, frameon=False)
    
    if(log):
        df[fields[2]][:] = np.log10(df[fields[2]][:])
    sns.kdeplot(np.nan_to_num(df[fields[2]][:], posinf=0.0), label="", color=color, shade=True, ax=ax[1,0])
    ax[1,0].axvline(np.nan_to_num(df[fields[2]][idx_city], posinf=0.0), linestyle='--', label=city)
    ax[1,0].set_xlabel(xlabels[2])
    ax[1,0].legend(loc='best', fontsize=24, frameon=False)

    sns.kdeplot(np.nan_to_num(df[fields[3]][:], posinf=0.0), label="", color=color, shade=True, ax=ax[1,1])
    ax[1,1].axvline(np.nan_to_num(df[fields[3]][idx_city], posinf=0.0), linestyle='--')
    ax[1,1].set_xlabel(xlabels[3])
    ax[1,1].legend(loc='best', fontsize=24, frameon=False)

    sns.kdeplot((df[fields[4]][:]), label="", color=color, shade=True, ax=ax[2,0])
    ax[2,0].axvline((df[fields[4]][idx_city]), linestyle='--', label=city)
    ax[2,0].set_xlabel(xlabels[4])
    ax[2,0].legend(loc='best', fontsize=24, frameon=False)


    sns.kdeplot((df[fields[5]][:]), label="", color=color, shade=True, ax=ax[2,1])
    ax[2,1].axvline((df[fields[5]][idx_city]), linestyle='--')
    ax[2,1].set_xlabel(xlabels[5])
    ax[2,1].legend(loc='best', fontsize=24, frameon=False)
    plt.show()
```

```python
def standardize_rank(arr, direction=1):
    rank = (stats.mstats.rankdata(np.ma.masked_invalid(arr)))
    rank[rank == 0] = np.nan
    if(direction == 1):
        return rank/(np.nanmax(rank)) 
    else:
        return 1-(rank/(np.nanmax(rank)))
```
