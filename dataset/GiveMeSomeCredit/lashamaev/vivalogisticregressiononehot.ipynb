```python
def chop(x):
    x = x.tolist()[0]
    coefficients = []
    for item in x:
        if np.abs(item)<0.001:
            coefficients.append(0.0)
        else:
            coefficients.append(round(item*10000)/10000)
    return coefficients        
            
for i in range(4):
    print(chop(classifiers[i].coef_))
```
