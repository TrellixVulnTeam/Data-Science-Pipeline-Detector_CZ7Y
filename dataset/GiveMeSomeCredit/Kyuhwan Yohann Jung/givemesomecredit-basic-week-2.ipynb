```python
# Outlier deleted
train = train[train['DebtRatio'] <50000]
```

```python
test = pd.read_csv('/kaggle/input/GiveMeSomeCredit/cs-test.csv')
test
```

```python
sub = pd.read_csv('/kaggle/input/GiveMeSomeCredit/sampleEntry.csv')
```

```python
alldata2 = alldata.drop(columns = ["Unnamed: 0" , 'SeriousDlqin2yrs' ])
alldata2
```

```python
train2 = alldata2[:len(train)]
test2 = alldata2[len(train):]
```

# RF, XGB, LGBM 

```python
# from sklearn.ensemble import RandomForestRegressor

# model = RandomForestRegressor(n_jobs = -1)

# from xgboost import XGBRegressor

# model = XGBRegressor()

from lightgbm import LGBMClassifier

model = LGBMClassifier()
```

```python
model.fit(train2, train['SeriousDlqin2yrs'])
```

```python
# Regressor --> model.predict
result = model.predict_proba(test2)
result
```
