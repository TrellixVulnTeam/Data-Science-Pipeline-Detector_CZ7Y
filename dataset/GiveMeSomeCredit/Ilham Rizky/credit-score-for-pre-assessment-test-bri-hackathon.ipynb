# Pertanyaan No. 14
Bangunlah sebuah model binomial logistic regression untuk memodelkan variabel Class dengan variabel prediktor berikut: age, DebtRatio, MonthlyIncome, NumberOfOpenCreditLinesAndLoans dan RevolvingUtilizationOfUnsecuredLines.

Berdasarkan tes signifikansi dari masing-masing prediktor, variabel manakah yang disimpulkan sebagai variabel signifikan terhadap peluang nasabah mengalami keterlambatan pembayaran di atas 90 hari
- [ ] DebtRatio
- [X] age
- [ ] RevolvingUtilizationOfUnsecuredLines
- [ ] NumberOfOpenCreditLinesAndLoans

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
columns = ['age', 
           'DebtRatio', 
           'MonthlyIncome',
           'NumberOfOpenCreditLinesAndLoans',
           'RevolvingUtilizationOfUnsecuredLines',
           'SeriousDlqin2yrs']
data = df[columns]
data['MonthlyIncome'] = data['MonthlyIncome'].fillna(0)
X = data.drop(['SeriousDlqin2yrs'], axis = 1)
y = data['SeriousDlqin2yrs']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .2, random_state=0)
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
coef = model.coef_[0]

pd.Series(coef, index = columns[:-1])
```
