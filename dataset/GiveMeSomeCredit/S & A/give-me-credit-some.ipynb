# MODELS

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import plot_confusion_matrix
```

```python
df = pd.read_csv("./train.csv")
X = df.drop(columns=["SeriousDlqin2yrs"])
Y = df["SeriousDlqin2yrs"]
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold,cross_val_score,GridSearchCV
from sklearn.preprocessing import StandardScaler
X = StandardScaler().fit_transform(X)
X,x,Y,y = train_test_split(X,Y,test_size=0.15)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression,LogisticRegressionCV
logit = LogisticRegression(random_state=111, solver='saga', penalty='l1', class_weight='unbalanced', C=1.0)
logit.fit(X,Y)
"Test Score :",logit.score(x,y) ,"Training Score :", logit.score(X,Y) 
```

```python
lr_pred = logit.predict(x)
lr_pred_proba = logit.predict_proba(x)
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score,accuracy_score,mean_absolute_error,confusion_matrix
# pred = model.predict_proba
rfc_params = {
    'n_estimators': list(range(10,101,10)),
    'max_depth': [13,17,19],
    'min_samples_leaf': [2, 4]
}
rfc_search =  GridSearchCV(RandomForestClassifier(),param_grid=rfc_params)
rfc_search.fit(X,Y)
print(rfc_search.best_score_ , rfc_search.best_params_)
```

```python
rfc = RandomForestClassifier(max_depth=13,min_samples_leaf= 4, n_estimators= 80).fit(X,Y)
rfc_pred = rfc_search.predict(x)
rfc_pred_proba = rfc_search.predict_proba(x)
plot_confusion_matrix(rfc,X,Y)
```

# Random Forest Regression

```python
rfr = RandomForestRegressor(max_depth= 9, min_samples_leaf= 2, n_estimators= 27).fit(X,Y)
rfr_pred = rfc.predict(X)
rfr_pred_proba = rfr.predict(x)
print(f1_score(Y,rfr_pred))
plot_confusion_matrix(rfc,X,Y)
```

# LGBMClassifier

```python
from lightgbm import LGBMClassifier
lgbm_params = {
    'n_estimators': [9,27,36],
    'max_depth': [3,7,9],
    'min_samples_leaf': [2, 4],
    
    'num_iterations': [100],
    'learning_rate': [0.005, 0.01],
    'n_estimators': [8,16,24],
    'num_leaves': [6,8,12,16],}
lgbm_search =  GridSearchCV(LGBMClassifier(),param_grid=lgbm_params,verbose=-1)
lgbm_search.fit(X,Y)
print("Best Score :",lgbm_search.best_score_ ,"Best Parameters:", lgbm_search.best_params_)
```

```python
lgbm = LGBMClassifier(learning_rate= 0.01, max_depth= 9, min_samples_leaf= 2, n_estimators= 8, num_iterations= 100, num_leaves= 16).fit(X,Y)
lgbm_pred = lgbm.predict(X)
lgbm_pred_proba = lgbm.predict_proba(x)
print("F1 Score :",f1_score(Y,lgbm_pred))
plot_confusion_matrix(lgbm,X,Y)
```

# XGBOOST Classifier

```python
from xgboost import XGBClassifier
parameters = {
    'max_depth': range (3, 22, 2),
    'n_estimators': list(range(10,101,10)),#range(60, 220, 40),
    'learning_rate': [0.1]#, 0.01, 0.05
}
XGBC = GridSearchCV(
    XGBClassifier(),
    param_grid=parameters,
    scoring = 'roc_auc',
    n_jobs = 10,
    cv = 2,
    verbose=True
)
XGBC.fit(X, Y)
print("Best Score :",XGBC.best_score_ ,"Best Params:",XGBC.best_params_)
pred =XGBC.predict(x)
print("F1 Score :",f1_score(y,pred))
```

```python
xgbc_model = XGBClassifier(learning_rate= 0.1, max_depth= 3, n_estimators= 90).fit(X,Y)
xgb_pred =xgbc_model.predict(x)
xgb_pred_proba = xgbc_model.predict_proba(x)
```
