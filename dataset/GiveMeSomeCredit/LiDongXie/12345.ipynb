```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame

train_data = pd.read_csv('/kaggle/input/GiveMeSomeCredit/cs-training.csv', index_col=False)
test_data = pd.read_csv('/kaggle/input/GiveMeSomeCredit/cs-test.csv', index_col=False)
all_data = train_data.append(test_data, sort=False).reset_index(drop=True)
all_data.drop(all_data.columns[[0]], axis=1, inplace=True)
#填充NA
all_data['MonthlyIncome'] = all_data['MonthlyIncome'].mask(all_data['MonthlyIncome'].isnull(), all_data['DebtRatio'])
numcols = all_data.columns
cols = numcols[numcols != 'SeriousDlqin2yrs']
for c in cols:
    all_data[c][all_data[c] == 'NA'] = np.NAN
for c in cols:
    all_data[c].fillna(0, inplace=True)
#特征工程
all_data.loc[all_data['age'] < 20, 'AgeType'] = 0
all_data.loc[(all_data['age'] >= 20) & (all_data['age'] < 50), 'AgeType'] = 1
all_data.loc[all_data['age'] > 50, 'AgeType'] = 2
all_data.loc[all_data['NumberOfTime30-59DaysPastDueNotWorse'] > 20, 'NumberOfTime30-59DaysPastDueNotWorse'] = 20
all_data.loc[all_data['NumberOfOpenCreditLinesAndLoans'] > 40, 'NumberOfOpenCreditLinesAndLoans'] = 40
all_data.loc[all_data['NumberOfTimes90DaysLate'] > 18, 'NumberOfTimes90DaysLate'] = 18
all_data.loc[all_data['NumberRealEstateLoansOrLines'] > 26, 'NumberRealEstateLoansOrLines'] = 26
all_data.loc[all_data['NumberOfTime60-89DaysPastDueNotWorse'] > 16, 'NumberOfTime60-89DaysPastDueNotWorse'] = 16
all_data.loc[all_data['NumberOfDependents'] > 8, 'NumberOfDependents'] = 8
#all_data['PerIncome'] = all_data['NumberOfOpenCreditLinesAndLoans'] / (all_data['RevolvingUtilizationOfUnsecuredLines'])
#all_data['PerDisposableIncome'] = all_data['DisposableIncome'] / (all_data['NumberOfDependents'] + 1)
all_data['PerIncome'] = all_data['MonthlyIncome'] / (all_data['DebtRatio'] + 0.0001)

#归一化
cols = numcols[(numcols != 'RevolvingUtilizationOfUnsecuredLines') & (numcols != 'DebtRatio')]
all_data = all_data.astype("float32")
all_data[cols] = (all_data[cols] - all_data[cols].min()) / (all_data[cols].max() - all_data[cols].min())
#划分训练集和测试集
X_train = all_data[all_data['SeriousDlqin2yrs'].notnull()].reset_index(drop=True)
X_test = all_data[all_data['SeriousDlqin2yrs'].isnull()].drop(['SeriousDlqin2yrs'], axis=1).reset_index(drop=True)
#划分训练集测试集
X_train.to_csv('train_data.csv', index=False)
X_test.to_csv('test_data.csv', index=False)
```

```python
train_data = pd.read_csv('train_data.csv')
test_data = pd.read_csv('test_data.csv')
#获取标签
label = train_data['SeriousDlqin2yrs']
#删除标签
train_data.drop(['SeriousDlqin2yrs'], axis=1, inplace=True)
#转换格式
train_data = np.array(train_data)
test_data = np.array(test_data)
label = np.array(label, 'int8')
#建立模型
automl = AutoML(
    mode="Compete", 
    eval_metric="auc",
    total_time_limit=6*3600,
    features_selection=False # switch off feature selection
)

automl.fit(train_data, label)
```
