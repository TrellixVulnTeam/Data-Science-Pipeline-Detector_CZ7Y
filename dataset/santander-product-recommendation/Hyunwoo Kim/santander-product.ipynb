# Data Visualization

```python
skip_cols = ['ncodpers', 'renta']
for col in df_train.columns:
    # 출력에 너무 시간이 오래 걸려서 ncodpers, renta는 제외
    if col in skip_cols:
        continue
    
    # 보기 편하게 변수명과 함께 출력한다.
    print('col : ', col)
    
    # 그래프
    f, ax = plt.subplots(figsize=(20, 15))
    # seaborn을 사용한 막대 그래프 생성
    sns.countplot(x=col, data=df_train, alpha=0.5)
    # show!
    plt.show()
```
