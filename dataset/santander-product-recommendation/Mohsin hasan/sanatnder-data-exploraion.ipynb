```python
#Lets check file sizes
def get_line_count(fnamepath):
    lcount = 0
    with open(fnamepath) as f:
        for line in f:
            lcount +=1   
    return lcount
```

```python
#Get column names from file
def get_colnames(fnamepath):
    with open(fnamepath) as f:
        colnames = f.readline().split()
    return colnames
```

**For visualization and understanding purposes, we will work only with 50% of data**

Lets read all customer codes in training data

```python
cust_codes = []
with open('../input/train_ver2.csv') as f:
    reader = csv.DictReader(f)
    for row in reader:
        if row['ncodpers'] not in cust_codes:
            cust_codes.append(row['ncodpers'])
```

Now, lets sample 50% of customer codes

```python
cust_codes_sample = np.random.choice(np.array(cust_codes), size = 0.5*len(cust_codes), replace=False)
```

**Now lets read train data corresponding to above customer codes**

```python
train_sample_data = []
with open('../input/train_ver2.csv') as f:
    reader = csv.DictReader(f)
    for row in reader:
        if row['ncodpers'] in cust_codes_sample:
            train_sample_data.append(row)
            
```

```python
def get_unique(fname, col_index):
    col_uni = []
    with open(fname) as f:
        for row in csv.reader(f):
            if row[col_index] not in col_uni:
                col_uni.append(row[col_index])
    return set(col_uni)
        
```
