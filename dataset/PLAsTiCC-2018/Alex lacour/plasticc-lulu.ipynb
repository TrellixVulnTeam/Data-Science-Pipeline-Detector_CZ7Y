```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

training_set_file = '/kaggle/input/PLAsTiCC-2018/training_set.csv'
training_set = pd.read_csv(training_set_file)  

training_set_meta_file = '/kaggle/input/PLAsTiCC-2018/training_set_metadata.csv'
training_set_meta = pd.read_csv(training_set_meta_file)  

 
```

```python
def display_object_id(object_id,training_set,training_set_meta) :
    print("+------------+ display object : {} ".format(object_id))
    print(training_set[training_set['object_id']==object_id])

    objdf = training_set[(training_set['object_id']==object_id)]
    meta = training_set_meta[training_set_meta['object_id']==object_id]

    fig, axs = plt.subplots(6)
    fig.suptitle('Object id:{} - target: {}'.format(object_id,meta['target']))
    
    for passband in [0,1,2,3,4,5] :
        objdf_passband = objdf[objdf['passband']==passband]
        x1 = objdf_passband['mjd']
        y1 = objdf_passband['flux']

        axs[passband].plot(x1, y1, label = "pb: {}".format(passband))
    plt.show(block = False)
```

```python
  
# Extract 10 object with same target
target = 90
sub_training_set_meta = training_set_meta[training_set_meta['target'] == target]

subset = 5
# sub_meta = sub_training_set_meta[0:subset] 
sub_meta = sub_training_set_meta.sample(n=subset)
# print(sub_meta)

for (indice, meta) in sub_meta.iterrows():
    #     print("meta : ",meta)
    display_object_id(meta['object_id'],training_set,training_set_meta)
```
