### 1. training data corner plots

```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd

data = pd.read_csv('../input/PLAsTiCC-2018/training_set_metadata.csv')
data_z = data[['hostgal_specz','hostgal_photoz']]
# log_data_z = pd.concat([np.log(data_z['hostgal_specz']),np.log(data_z['hostgal_photoz'])], axis = 1)
fig = corner.corner(data_z,range=[(0,2.4),(0,2.4)], show_titles=True, title_kwargs={"fontsize":10},label_kwargs = {'fontsize':14})
for ax in fig.get_axes():   
      ax.tick_params(axis='both', labelsize=14)

del data_z
```

### 2. training data AGN corner plots

```python
data = pd.read_csv('../input/PLAsTiCC-2018/training_set_metadata.csv')
agn_data = data[data.target==88]
data_z = agn_data[['hostgal_specz','hostgal_photoz']]
log_data_z = pd.concat([np.log10(data_z['hostgal_specz']),np.log10(data_z['hostgal_photoz'])], axis = 1)
fig = corner.corner(data_z, labels = ['log(specz)','log(photoz)'],show_titles=True, title_kwargs={"fontsize":10},label_kwargs = {'fontsize':14})
for ax in fig.get_axes():   
      ax.tick_params(axis='both', labelsize=14)
del data_z
```

```python
data_photoz = agn_data[['hostgal_photoz','hostgal_photoz_err']]
figure = corner.corner(data_photoz, show_titles=True, title_kwargs={"fontsize": 12})
del data_photoz
```
