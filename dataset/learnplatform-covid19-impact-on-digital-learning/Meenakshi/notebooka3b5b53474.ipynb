```python
dis_int={'district_id':'int64'}
engagement=engagement.astype(dis_int)
engagement
```

```python
engagement['time']=pd.to_datetime(engagement['time'])
```

```python
mapping_state=dict(district_info[['district_id','state']].values)
mapping_local= dict(district_info[['district_id','locale']].values)
engagement['state']=engagement.district_id.map(mapping_state)
engagement['local']=engagement.district_id.map(mapping_local)
```

```python
engagement.head()
```

```python
engagement.isnull().sum()
```

```python
engagement_state=engagement[['pct_access','engagement_index','state']]
engagement_local= engagement[['pct_access','engagement_index','local']]
```

```python
plt.figure(figsize=(15,10))
plt.ticklabel_format(style='plain')
sns.countplot(y='state',data= engagement_state,order=engagement_state.state.value_counts().index,palette='rocket_r',linewidth=4)
plt.title("statewise digital connectivity in 2020",font='Georgia',size=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('count',fontsize=12)
plt.ylabel('state',fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(12,10))
plt.ticklabel_format(style='plain')
sns.countplot(y='local',data= engagement_local,order=engagement_local.local.value_counts().index,palette='mako',linewidth=3)
plt.title('local-wise digital connectivity distribution in 2020',font='Georgia',size=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('count',size=12)
plt.ylabel('local',size=12)
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(15,10))
explode=(0.04,0.04,0.04,0.04)
labels=list(engagement_local.local.value_counts().index)
sizes=engagement_local.local.value_counts().values
patches,texts,autotexts= ax.pie(sizes,explode=explode,startangle=60,labels=labels,
                                autopct='%1.0f%%',pctdistance=0.7,
                               colors=['#367382','#3998af','#16b5dc','#a8c2c9'])
texts[0].set_fontsize(20)
texts[1].set_fontsize(20)
texts[2].set_fontsize(20)
texts[3].set_fontsize(20)
autotexts[0].set_fontsize(20)
autotexts[1].set_fontsize(20)
autotexts[2].set_fontsize(20)
autotexts[3].set_fontsize(20)
ax.add_artist(plt.Circle((0,0),0.6,fc='white'))
font={'fontname':'Georgia'}
plt.title("local-wise digital conectivity distribution in 2020",fontsize=20,**font)
plt.show()
```
