# Reading districts data

```python
districts_data = pd.read_csv("../input/learnplatform-covid19-impact-on-digital-learning/districts_info.csv")
districts_data.head()
```

# Drop all Nan values in districts_data

```python
districts_data = districts_data.dropna()
districts_data
```

```python
districts_data.reset_index(inplace=True)
districts_data = districts_data.drop(["index"],axis=1)
districts_data
```

#  Unique values in each column

```python
unique = [districts_data[column].unique() for column in districts_data.columns]
unique
```

# Visualizing using matplotlib

```python
x = [len(districts_data[column].unique()) for column in districts_data.columns]
plt.figure(figsize=(10,10))
plt.xticks(rotation=-90)
plt.bar(districts_data.columns,x)
```

# Exploration in state

```python
plt.figure(figsize=(10,10))
plt.xticks(rotation=-90)
plt.bar(districts_data["state"].unique(),[int for int in districts_data["state"].value_counts()])
```

# Exploration in locale

```python
plt.figure(figsize=(10,10))
plt.xticks(rotation=-90)
plt.bar(districts_data["locale"].unique(),[int for int in districts_data["locale"].value_counts()])
```
