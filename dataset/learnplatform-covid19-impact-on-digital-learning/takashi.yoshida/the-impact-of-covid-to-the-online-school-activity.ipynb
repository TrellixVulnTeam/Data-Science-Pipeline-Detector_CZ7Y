# The impact of COVID to the online school activity
In this notebook, I will try to identify the variant that influence the student engagement rate. Below, I will prepare the data, visualize the relation between the various variables and student engagement rate and finally conclude what variable has large infuluence to the student engagement rate. 


### Step 1 : Data preparation
Here, from the engagement data I get mean of mean of pct_access per day and mean of engagement_index accross all prodact per day from each district. Also, from district data, I get several variables and conbine with engagement data.  

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

#get district ID, mean of mean of pct_access per day, and mean of engagement_index accross all prodact per day
engage_list = []
for dirname, _, filenames in os.walk('../input/learnplatform-covid19-impact-on-digital-learning/engagement_data/'):
    for filename in filenames:
        file_path = os.path.join(dirname, filename)
        X = pd.read_csv(file_path)
        pct_mean = X["pct_access"].mean()
        eng_mean = X["engagement_index"].sum()
        file_num = filename.replace('.csv', '')
        engage_list.append([file_num, pct_mean, eng_mean])
engage_df = pd.DataFrame(engage_list)
engage_df.columns = ["district_id", "pct_mean" , "eng_mean"]
engage_df['district_id'] = engage_df['district_id'].astype(int)

engage_df.head()
```

```python
#get data from district_info.csv
dist_df = pd.read_csv("../input/learnplatform-covid19-impact-on-digital-learning/districts_info.csv")
dist_df.head()
```

```python
#combine engagement data and dist data
df = pd.merge(engage_df, dist_df,on="district_id")
#This is the data I am going to deal with!
df.head()
```

```python
#make range data to numerical data!
df['pct_black/hispanic'] = df['pct_black/hispanic'].str.replace("[",'')#.find(',')
df['pct_black/hispanic'] = df['pct_black/hispanic'].str.split(',')
df['pct_black/hispanic'] = df['pct_black/hispanic'].str[0].astype(float) + 0.1
df['pct_free/reduced'] = df['pct_free/reduced'].str.replace("[",'')#.find(',')
df['pct_free/reduced'] = df['pct_free/reduced'].str.split(',')
df['pct_free/reduced'] = df['pct_free/reduced'].str[0].astype(float) + 0.1
df['pp_total_raw'] = df['pp_total_raw'].str.replace("[",'')#.find(',')
df['pp_total_raw'] = df['pp_total_raw'].str.split(',')
df['pp_total_raw'] = df['pp_total_raw'].str[0].astype(float) + 1000
df.loc[df['locale'] == 'Suburb', "locate_num"] = 1
df.loc[df['locale'] == 'City', "locate_num"] = 2
df.loc[df['locale'] == 'Rural', "locate_num"] = 0
```

### Step 2 Data Visualization
Plot all the data

```python
cpr = df.corr()
cpr["pct_mean"]
```

```python
fig, ax = plt.subplots(2,2,figsize=(15,10)) # 論理サイズ縦20 x 横15のマンションに、縦4 x 横3枠の部屋を確保

plt.subplots_adjust(wspace=0.4, hspace=0.4)

var_list = ["pct_black/hispanic","pct_free/reduced","pp_total_raw","locate_num"]
for i, var in enumerate(var_list):
    df.plot.scatter(x=var,y="pct_mean",ax=ax[i//2,i%2])
    ax[i//2,i%2].set_title("correlation between pct_mean and {}".format(var))

print("Correlation Coefficient with pct_mean")
print(df.corr()["pct_mean"])
print("\n")
```

```python
fig, ax = plt.subplots(2,2,figsize=(15,10)) # 論理サイズ縦20 x 横15のマンションに、縦4 x 横3枠の部屋を確保

plt.subplots_adjust(wspace=0.4, hspace=0.4)

var_list = ["pct_black/hispanic","pct_free/reduced","pp_total_raw","locate_num"]
for i, var in enumerate(var_list):
    df.plot.scatter(x=var,y="eng_mean",ax=ax[i//2,i%2])
    ax[i//2,i%2].set_title("correlation between eng_mean and {}".format(var))

print("Correlation Coefficient with pct_mean")
print(df.corr()["eng_mean"])
print("\n")
```

### Step 3 : Conclusion
From the analysis above, none of variant sgnificantly affect the percentage of participation or engagement rate. However, pp_total_raw can explain a little about the change of engagement rate.
So from this analysie, enhancing the access to degital devece is the most inpactful intervation to increase engagement rate of student!
