```python
def range_changer(col_name):
    '''
    input: column from the dataset with shape [range[
    return: the same column with shape [range]
    '''
    col = dist_data[col_name]
    new_col = []
    for i in range(len(col)):
        if str(col[i])!= "nan":
            dist_data[col_name].loc[i] = dist_data[col_name].loc[i][:-1]
            dist_data[col_name].loc[i]= dist_data[col_name].loc[i]+ "]"
```

```python
range_to_change= ["pct_black/hispanic", "pct_free/reduced", "county_connections_ratio",
                     "pp_total_raw"]
dist_data["pp_total_raw"].loc[0]
```

```python
for change_col in range_to_change:
    range_changer(change_col)
```

```python
obj_data= dist_data.select_dtypes('object')
for obj_col in obj_data:
    print("--------------------------------------------------------")    
    print(obj_col,"uniques:")
    print("--------------------------------------------------------")    
    unq= dist_data[obj_col].unique()
    for j in range(len(unq)):
        print(unq[j])
    print("\n")
```

```python
feature_name= []
null_count= []
for i in dist_data.columns:
    feature_name.append(i)
    null_count.append(dist_data[i].isnull().sum())
print(feature_name)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,2,1])
ax.bar(feature_name,null_count)
plt.title("Bar Chart that show how many null values each feature contains ",
         fontdict={'fontsize': 18, 'fontweight': 'bold', 'color':'#138FBD'})
plt.show()
```

```python
obj_data= ['Sector(s)', 'Primary Essential Function', 'Provider/Company Name']
for obj_col in obj_data:
    print("--------------------------------------------------------")    
    print(obj_col,"uniques:",(len(prod_data[obj_col].unique())))
    print("--------------------------------------------------------")    
    unq= prod_data[obj_col].unique()
    for j in range(len(unq)):
        print(unq[j])
    print("\n")
```

```python
PEF_1st=[] #Primary Essential Function first layer

#create a columns that change the values to LC, CM, SDO.
for i in prod_data['Primary Essential Function']:
    if i != i:
        PEF_1st.append("nan")
    elif (str(i[0])+str(i[1])) =="LC":
        PEF_1st.append("LC")
    elif (str(i[0])+str(i[1])) =="CM":
        PEF_1st.append("CM")
    elif (str(i[0])+str(i[1])) =="SD":
        PEF_1st.append("SDO")
        
PEF_1st = np.array(PEF_1st).reshape(372,1) #convert list to array and reshape it.
PEF_1st = pd.DataFrame(PEF_1st, columns=['PEF_1st']) # convert the array to data frame.
prod_data=pd.concat([prod_data,PEF_1st],axis=1) #merge the columns to the product dataset
```
