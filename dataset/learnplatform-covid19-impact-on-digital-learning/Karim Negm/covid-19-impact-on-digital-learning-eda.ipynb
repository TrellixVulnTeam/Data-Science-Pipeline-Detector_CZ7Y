## Data Wrangling


### the districts_info file description
- this  file includes information about the characteristics of school districts, including data from NCES and FCC.
- it has 233 sample with 7 features 

the features name is ['district_id' 'state' 'locale' 'pct_black/hispanic' 'pct_free/reduced'
 'county_connections_ratio' 'pp_total_raw']
 
 
- doesn't contain any duplicate values- 
- Have the follwing missing valuse:
       - district_id                 0
       -state                        57
       -locale                       57
       -pct_black/hispanic           57
       -pct_free/reduced             85
       -county_connections_ratio     71
       -pp_total_raw                115
