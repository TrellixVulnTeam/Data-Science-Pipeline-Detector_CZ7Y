## Engagement Data


```python
path = '../input/learnplatform-covid19-impact-on-digital-learning/engagement_data'
```

```python
import glob
file = glob.glob(path+ '/*.csv')
lst = []
for f in file:
    df = pd.read_csv(f, index_col = None, header = 0)
    lst.append(df)
```

```python
eng_df = pd.concat(lst, axis = 0)
```

## Distribution of Sector(s) Data

```python

ds = df_prod['Sector(s)'].value_counts().reset_index()
ds.columns = ['Sector(s)'
             ,'percentage']
ds["percentage"] /= len(df_prod)
ds
```

```python
import plotly.express as px
fig = px.pie(ds, 
            names = 'Sector(s)',
            values = 'percentage',
            color_discrete_sequence = px.colors.sequential.ice,
            title = "Distribution of Sectors Data")
fig.show()
```

## Occurence of states in the Districts Information Data

```python
df_dist
```

```python
state_occur = df_dist['state'].value_counts().reset_index()
state_occur.columns = ['state'
             ,'percentage']
state_occur["percentage"] /= len(df_dist)
state_occur
```

```python
fig = px.pie(state_occur, 
            names = 'state',
            values = 'percentage',
            color_discrete_sequence = px.colors.sequential.Plasma,
            title = "Distribution of Occurence of state")
fig.show()
```

## Occurnece of locale

```python
loc_occur = df_dist['locale'].value_counts().reset_index()
loc_occur.columns = ['locale'
             ,'percentage']
loc_occur["percentage"] /= len(df_dist)
loc_occur
```

```python
fig = px.pie(loc_occur, 
            names = 'locale',
            values = 'percentage',
            color_discrete_sequence = px.colors.sequential.Burg,
            title = "Distribution of Occurence of state")
fig.show()
```
