```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
```

```python
districts = pd.read_csv('/kaggle/input/learnplatform-covid19-impact-on-digital-learning/districts_info.csv')
districts.head()
```

```python
# combine all district engagement data files into one data frame
engagement = pd.DataFrame()
for dirname, _, filenames in os.walk('/kaggle/input/learnplatform-covid19-impact-on-digital-learning/engagement_data/'):
    for filename in filenames:
        district_engagement = pd.read_csv(os.path.join(dirname, filename))
        district_engagement['district_id'] = float(filename.split('.')[0])
        engagement = engagement.append(district_engagement)
engagement.head()
```

```python
engagement.drop(['time', 'lp_id', 'pct_access'], axis=1, inplace=True)
engagement
```

```python
# Focus on average engagement index by district
engagement = engagement.groupby('district_id').mean()
```

```python
engagement = pd.merge(engagement, districts, on='district_id')
engagement.dropna(inplace=True)
engagement.head()
```

```python
engagement = engagement[['engagement_index', 'pct_black/hispanic', 'pct_free/reduced', 'pp_total_raw']]
engagement
```

```python
# Convert interval values to means

def get_interval_mean(interval):
    if str(interval) != 'nan':
        limits = str(interval).split(',')
        lower_limit = float(limits[0][1:])
        upper_limit = float(limits[1][:-1])
        return (lower_limit + upper_limit)/2
    else:
        return None

engagement['pct_black/hispanic_mean'] = engagement['pct_black/hispanic'].apply(
    get_interval_mean)
engagement['pct_free/reduced_mean'] = engagement['pct_free/reduced'].apply(
    get_interval_mean)
engagement['pp_total_raw_mean'] = engagement['pp_total_raw'].apply(
    get_interval_mean)
```

```python
engagement = engagement[['engagement_index', 'pct_black/hispanic_mean', 
                         'pct_free/reduced_mean', 'pp_total_raw_mean']]
engagement
```

```python
engagement.corr()
```

# Conclusions: 
There is a significant positive correlation (0.26) between per-pupil expenditure and average pupil engagement with online tools by district.

There is a significant negative correlation (-0.15) between percentage of Black/Hispanic students and average pupil engagement with online tools by district.
