⚙️Build a function that will change [range[ --> [range]

```python
def range_changer(col_name):
    '''
    input: column from the dataset with shape [range[
    return: the same column with shape [range]
    '''
    col = dist_data[col_name]
    new_col = []
    for i in range(len(col)):
        if str(col[i])!= "nan":
            dist_data[col_name].loc[i] = dist_data[col_name].loc[i][:-1]
            dist_data[col_name].loc[i]= dist_data[col_name].loc[i]+ "]"
```

```python
range_to_change= ["pct_black/hispanic", "pct_free/reduced", "county_connections_ratio",
                     "pp_total_raw"]
dist_data["pp_total_raw"].loc[0]
```

```python
for change_col in range_to_change:
    range_changer(change_col)
```

 Districts Data Visualization

```python
feature_name= []
null_count= []
for i in dist_data.columns:
    feature_name.append(i)
    null_count.append(dist_data[i].isnull().sum())
```

```python
print(feature_name)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,2,1])
ax.bar(feature_name,null_count)
plt.title("Bar Chart that show how many null values each feature contains ",
         fontdict={'fontsize': 18, 'fontweight': 'bold', 'color':'#138FBD'})
plt.show()
```

```python
dist_data.groupby(['state','locale'])['locale'].count().sort_values().plot.pie(figsize=(25,25))
plt.title("Pie chart shows the frequance of every Locale in each State ",
         fontdict={'fontsize': 18, 'fontweight': 'bold', 'color':'#13BD6F'})
```

```python
dist_data.groupby(['state','locale'])['state'].count().plot.barh(figsize=(15,15), color='#55FF06')
plt.title("Bar chart shows the frequance of every Locale in each State ",
         fontdict={'fontsize': 18, 'fontweight': 'bold', 'color':'#581845'})
```

```python
dist_data.groupby(['locale','county_connections_ratio'])['locale'].count().plot.pie(figsize=(20,10))
plt.title("Pie chart that show the county connections ration of each locale ",
         fontdict={'fontsize': 18, 'fontweight': 'bold', 'color':'#AAA119'})
```

```python
dist_data.groupby(['pct_black/hispanic'])['locale'].value_counts().sort_index().plot.pie(figsize=(15,15), autopct="%1.1f%%")
plt.title("Pie chart that show the count of the ratio of black or hispanic students in each locale",
         fontdict={'fontsize': 18, 'fontweight': 'bold', 'color':'#FFC300'})
```

```python
dist_data.groupby(['state','locale'])['pct_black/hispanic'].value_counts().sort_index().plot.bar(figsize=(25,8))
plt.title("Bar chart that show the count of the ratio of black or hispanic students in each locale of each state",
         fontdict={'fontsize': 18, 'fontweight': 'bold', 'color':'blue'})
```

```python
dist_data.groupby(['state','locale'])['pp_total_raw'].value_counts().sort_index().plot.bar(figsize=(20,8), color="#FF8606")
plt.title("Bar chart that show the count of the school expenditure for each locale of each state",
         fontdict={'fontsize': 18, 'fontweight': 'bold', 'color':'#FF5733'})
```

```python
PEF_1st=[] #Primary Essential Function first layer

#create a columns that change the values to LC, CM, SDO.
for i in prod_data['Primary Essential Function']:
    if i != i:
        PEF_1st.append("nan")
    elif (str(i[0])+str(i[1])) =="LC":
        PEF_1st.append("LC")
    elif (str(i[0])+str(i[1])) =="CM":
        PEF_1st.append("CM")
    elif (str(i[0])+str(i[1])) =="SD":
        PEF_1st.append("SDO")
        
PEF_1st = np.array(PEF_1st).reshape(372,1) #convert list to array and reshape it.
PEF_1st = pd.DataFrame(PEF_1st, columns=['PEF_1st']) # convert the array to data frame.
prod_data=pd.concat([prod_data,PEF_1st],axis=1) #merge the columns to the product dataset
```
