```python
district_df.dropna(subset = ["state"], inplace=True)
district_df = district_df.reset_index(drop=True)
district_df.head()
```

```python
district_df['pct_free/reduced'] = district_df['pct_free/reduced'].fillna(district_df['pct_free/reduced'].mode()[0])
district_df['county_connections_ratio'] = district_df['county_connections_ratio'].fillna(district_df['county_connections_ratio'].mode()[0])
district_df['pp_total_raw'] = district_df['pp_total_raw'].fillna(district_df['pp_total_raw'].mode()[0])
```

```python
district_df.isnull().sum()
```

```python
def change_to_percentage(string):
    if string == '[0, 0.2[':
        return 0.1
    elif string == '[0.2, 0.4[':
        return 0.3
    elif string == '[0.4, 0.6[':
        return 0.5
    elif string == '[0.6, 0.8[':
        return 0.7
    elif string == '[0.8, 1[':
        return 0.9
    else:
        pass

district_df['pct_black/hispanic'] = district_df['pct_black/hispanic'].apply(lambda x: change_to_percentage(x))
district_df['pct_free/reduced'] = district_df['pct_free/reduced'].apply(lambda x: change_to_percentage(x))
```

```python
def change_conn_ratio(string):
    if string == '[0.18, 1[':
        return 'low-speed'
    elif string == '[1, 2[':
        return 'high-speed'
    else:
        pass

district_df['county_connections_ratio'] = district_df['county_connections_ratio'].apply(lambda x: change_conn_ratio(x))
```

```python
def change_sum_expenditure(string):
    if string == '[4000, 6000[':
        return 5000
    elif string == '[6000, 8000[':
        return 7000
    elif string == '[8000, 10000[':
        return 9000
    elif string == '[10000, 12000[':
        return 11000
    elif string == '[12000, 14000[':
        return 13000
    elif string == '[14000, 16000[':
        return 15000
    elif string == '[16000, 18000[':
        return 17000
    elif string == '[18000, 20000[':
        return 19000
    elif string == '[20000, 22000[':
        return 21000
    elif string == '[22000, 24000[':
        return 23000
    elif string == '[32000, 34000[':
        return 33000
    else:
        pass

district_df['pp_total_raw'] = district_df['pp_total_raw'].apply(lambda x: change_sum_expenditure(x))
```

```python
engagement = []
for district_id in district_df["district_id"]:
    df = pd.read_csv(r'../input/learnplatform-covid19-impact-on-digital-learning/engagement_data/'+str(district_id)+'.csv')
    df.insert(1, 'district_id', district_id)
    engagement.append(df)
    
engagement_df = pd.concat(engagement)
engagement_df = engagement_df.reset_index(drop=True)
```

```python
engagement_df.head()
```

```python
engagement_df['time'] = pd.to_datetime(engagement_df['time'])
```

```python
engagement_df.info()
```

```python
engagement_df.isnull().sum()
```

```python
engagement_df.dropna(subset = ["engagement_index"], inplace=True)
engagement_df.dropna(subset = ["lp_id"], inplace=True)
engagement_df = engagement_df.reset_index(drop=True)
engagement_df.head()
```

```python
engage_max_access = engagement_df.groupby(['district_id', 'time']).pct_access.max()
engage_max_access
```

```python
engage_sum_index = engagement_df.groupby(['district_id', 'time']).engagement_index.sum()
engage_sum_index
```

```python
group_engagement_df = pd.concat([engage_max_access, engage_sum_index], axis=1)
group_engagement_df
```
