```python
dr=pd.read_csv('../input/learnplatform-covid19-impact-on-digital-learning/districts_info.csv')
dr.head()
```

```python
dr.info()
```

```python
ddr=dr
ddr.head()
```

```python
dr[dr.isnull().any(axis=1)]
```

```python
ddr.dropna(subset = ["state","locale"], inplace=True)
ddr
```

```python
ddr[['pct_black/hispanic','pct_free/reduced','county_connections_ratio','pp_total_raw']]=ddr[['pct_black/hispanic','pct_free/reduced','county_connections_ratio','pp_total_raw']].replace(np.nan,"")
ddr.head()
```

```python
ddr.info()
```

```python
 for col in ddr[['pct_black/hispanic','pct_free/reduced','county_connections_ratio','pp_total_raw']]:
    ddr[col] = ddr[col].map(lambda x: x.replace('[',''))
    ddr[col] = ddr[col].map(lambda x: x.replace(',',''))
ddr
```

```python
ddr[['pct_black/hispanic0','pct_black/hispanic1']]=ddr['pct_black/hispanic'].str.split(expand=True)
ddr[['pct_free/reduced0','pct_free/reduced1']]=ddr['pct_free/reduced'].str.split(expand=True)
ddr[['county_connections_ratio0','county_connections_ratio1']]=ddr['county_connections_ratio'].str.split(expand=True)
ddr[['pp_total_raw0','pp_total_raw1']]=ddr['pp_total_raw'].str.split(expand=True)
```

```python
ddr.head()
```

```python
ddr.drop(['pct_black/hispanic','pct_free/reduced','county_connections_ratio','pp_total_raw'],axis=1,inplace=True)
```

```python
ddr[['pct_black/hispanic0','pct_black/hispanic1','pct_free/reduced0','pct_free/reduced1','county_connections_ratio0','county_connections_ratio1']]=ddr[['pct_black/hispanic0','pct_black/hispanic1','pct_free/reduced0','pct_free/reduced1','county_connections_ratio0','county_connections_ratio1']].apply(pd.to_numeric)
```

```python
ddr.info()
```

```python
ddr[['pct_black/hispanic0','pct_black/hispanic1','pct_free/reduced0','pct_free/reduced1','county_connections_ratio0','county_connections_ratio1']]=ddr[['pct_black/hispanic0','pct_black/hispanic1','pct_free/reduced0','pct_free/reduced1','county_connections_ratio0','county_connections_ratio1']].multiply(100)
```
