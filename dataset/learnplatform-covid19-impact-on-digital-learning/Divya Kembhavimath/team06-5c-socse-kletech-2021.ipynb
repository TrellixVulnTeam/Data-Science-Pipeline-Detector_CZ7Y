**PREPROCESSING**

```python
path = '/kaggle/input/learnplatform-covid19-impact-on-digital-learning/engagement_data' 
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    district_id = filename.split("/")[4].split(".")[0]
    x = district_id[16:]
    df["district_id"] = x
    li.append(df)
    
engagement_df = pd.concat(li)
engagement_df = engagement_df.reset_index(drop=True)
engagement_df
```

```python
engagement_df.isnull().sum(axis = 0)
```

```python
total = engagement_df.isnull().sum().sort_values(ascending=False)
percent = (engagement_df.isnull().sum()/engagement_df.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(20)
```

```python
#After handling missing data, we will cast a couple of columns in the unified engagements dataframe (lp_id, district_id) to int. 
#It will be helpful down the road as we are going to merge engagements data with district and product information.
eng_df = engagement_df.drop(engagement_df.loc[engagement_df['lp_id'].isnull()].index)
eng_df = engagement_df.fillna('--unknown--')
```
