```python
df = pd.read_csv("/kaggle/input/learnplatform-covid19-impact-on-digital-learning/products_info.csv")
df1 = pd.read_csv("/kaggle/input/learnplatform-covid19-impact-on-digital-learning/districts_info.csv")
```

```python
import matplotlib.pyplot as plt
df_null = pd.DataFrame(df.isnull().sum()/len(df)).reset_index().sort_values(by=0)
plt.figure(figsize=(8,6))
plt.bar(df_null['index'],df_null[0],alpha=0.75)
plt.xticks(rotation=90)
plt.rc('ytick', labelsize=10)
plt.title('Percetage of missing values in products data')
```

```python
df1_null = pd.DataFrame(df1.isnull().sum()/len(df1)).reset_index().sort_values(by=0)
plt.figure(figsize=(8,6))
plt.bar(df1_null['index'],df1_null[0],alpha=0.75)
plt.xticks(rotation=90)
plt.rc('ytick', labelsize=10)
plt.title('Percetage of missing values in products data')
```
