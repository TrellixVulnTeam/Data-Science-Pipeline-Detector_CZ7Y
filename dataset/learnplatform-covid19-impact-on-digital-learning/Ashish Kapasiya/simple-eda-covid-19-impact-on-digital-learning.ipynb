```python
district.rename(columns={'pct_black/hispanic':'pct_black_hispanic',
                         'pct_free/reduced':'pct_free_reduced'}, inplace=True)
```

```python
district['pct_black_hispanic'] = district['pct_black_hispanic'].map(lambda x: str(x)[1:-1])
district['pct_free_reduced'] = district['pct_free_reduced'].map(lambda x: str(x)[1:-1])
district['pp_total_raw'] = district['pp_total_raw'].map(lambda x: str(x)[1:-1])
district['county_connections_ratio'] = district['county_connections_ratio'].map(lambda x: str(x)[1:-1])
```

```python
district["pp_total_raw_max"]= district["pp_total_raw"].str.split(",", n = 1, expand = True)[1] 
district["pp_total_raw_min"]= district["pp_total_raw"].str.split(",", n = 1, expand = True)[0] 
district["pct_black_hispanic_max"]= district["pct_black_hispanic"].str.split(",", n = 1, expand = True)[1] 
district["pct_black_hispanic_min"]= district["pct_black_hispanic"].str.split(",", n = 1, expand = True)[0]
district["pct_free_reduced_max"]= district["pct_free_reduced"].str.split(",", n = 1, expand = True)[1] 
district["pct_free_reduced_min"]= district["pct_free_reduced"].str.split(",", n = 1, expand = True)[0]
district["county_connections_ratio_max"]= district["county_connections_ratio"].str.split(",", n = 1, expand = True)[1] 
district["county_connections_ratio_min"]= district["county_connections_ratio"].str.split(",", n = 1, expand = True)[0]
```

```python
district["pp_total_raw_max"] = pd.to_numeric(district["pp_total_raw_max"],errors = 'coerce')
district["pp_total_raw_min"] = pd.to_numeric(district["pp_total_raw_min"],errors = 'coerce')
district["pct_black_hispanic_max"] = pd.to_numeric(district["pct_black_hispanic_max"],errors = 'coerce')
district["pct_black_hispanic_min"] = pd.to_numeric(district["pct_black_hispanic_min"],errors = 'coerce')
district["pct_free_reduced_max"] = pd.to_numeric(district["pct_free_reduced_max"],errors = 'coerce')
district["pct_free_reduced_min"] = pd.to_numeric(district["pct_free_reduced_min"],errors = 'coerce')
district["county_connections_ratio_max"] = pd.to_numeric(district["county_connections_ratio_max"],errors = 'coerce')
district["county_connections_ratio_min"] = pd.to_numeric(district["county_connections_ratio_min"],errors = 'coerce')
```

```python
district.drop(['pct_black_hispanic','pct_free_reduced','county_connections_ratio','pp_total_raw'],axis=1,inplace=True)
```

```python
district.head()
```

```python
district['state'] = district['state'].fillna('unknown')
```

```python
district['locale'].value_counts()
```

```python
district['locale'] = district['locale'].fillna('other')
```

```python
district.isnull().sum()
```

```python
district['pp_total_raw_min'].value_counts()
```

```python
district['pp_total_raw_max'].value_counts()
```

```python
sum0= district.pp_total_raw_max + district.pp_total_raw_min
district['avg_total_raw'] = (sum0 / 2)
sum1 = district.pct_black_hispanic_max + district.pct_black_hispanic_min
district['avg_black_hispanic'] = (sum1 / 2)
sum2 = district.pct_free_reduced_max + district.pct_free_reduced_min
district['avg_free_reduced'] = (sum2 / 2)
```

```python
district['county_connections_ratio_max'].value_counts()
```

```python
district['county_connections_ratio_min'] = district['county_connections_ratio_min'].replace('a',1)
```

```python
district['county_connections_ratio_min'] = district['county_connections_ratio_min'].fillna(0.18)
district['county_connections_ratio_max'] = district['county_connections_ratio_max'].fillna(2.0)
```

```python
district['avg_total_raw'] = district['avg_total_raw'].fillna(np.mean(district['avg_total_raw']))
district['avg_black_hispanic'] = district['avg_black_hispanic'].fillna(np.mean(district['avg_black_hispanic']))
district['avg_free_reduced'] = district['avg_free_reduced'].fillna(np.mean(district['avg_free_reduced']))
```

```python
district.drop(['pp_total_raw_max',
       'pp_total_raw_min', 'pct_black_hispanic_max', 'pct_black_hispanic_min',
       'pct_free_reduced_max', 'pct_free_reduced_min'],axis=1,inplace=True)
```
