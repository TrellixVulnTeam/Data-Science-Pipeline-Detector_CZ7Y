```python
import re

def get_lower_range(p_string):
    """
    get_lower_range splits range values and returns lower part of it.
    The value is in format "[<lower_end>, <upper_end>[". For example "[0.18, 1[", in which case "0.18" will be returned.
    
    Parameters:
        p_string - string to process, for example "[0.18, 1["
        
    Example:
        >>>> s = get_lower_range("[0.18, 1[")
        "0.18"
    """
    
    if pd.isnull(p_string):
        return p_string
    
    ret = re.search('\[.+,', p_string)
    ret = ret.group().replace('[', '').replace(',', '')
    ret = float(ret)

    return ret

def get_upper_range(p_string):
    """
    get_upper_range splits range values and returns upper part of it.
    The value is in format "[<lower_end>, <upper_end>[". For example "[0.18, 1[", in which case "1" will be returned.
    
    Parameters:
        p_string - string to process, for example "[0.18, 1["
        
    Example:
        >>>> s = get_upper_range("[0.18, 1[")
        "1"
    """
    
    if pd.isnull(p_string):
        return p_string
    
    ret = re.search(',.+\[', p_string)
    ret = ret.group().replace(', ', '').replace('[', '')
    ret = float(ret)

    return ret
```

```python
# Expand categories list into one-hot columns

def expand_categories_one_hot(df, col_name, sep=';', replace_str=' ', replace_with='', pref=None):
    """
    expand_categories_one_hot creates one-hot categories out of categorical values of a given column
        The categories in the column may be represented by compound values, concatenated with a separator. For example "PreK-12; Higher Ed", in
    which case they can be split into two corresponding columns: "PreK-12" and "Higher Ed".
    Also, the values in the column can be pre-processed by replacing specific string. For example, removing space symbol from "PreK-12; Higher Ed"
    before transforming to one-hot attributes.
    
    Parameters:
        df - Source DataFrame
        col_name - Column name in df, which is used for expansion
        sep - Symbol-separator between values. Default ";"
        replace_str - String to replace, on the pre-processing step. Default " " (space)
        replace_with - String to replace with, on the pre-processing step. Default "" (empty string)
        pref - Prefix of new column names. By default, the column names will become the same as the categorical values. pref will add prefix to each of
                these columns
                
    Returns:
        A DataFrame same as original one with added one-hot encoded columns
                
    Example:
        # In the example below we take Products data (represented as df_products DataFrame) and make one-hot encoding from its 'Sector(s)' column.
        # The 'Sector(s)' will remain intact.
        # On a pre-processing step we remove spaces 
        >>>> df_p = expand_categories_one_hot(df_products, 'Sector(s)', replace_str='; ', replace_with=';', pref='sec_')
    """
    
    stg = df[col_name]
    
    if replace_str:
        stg = stg.str.replace(replace_str, replace_with)
        
    stg = stg.str.get_dummies(sep)

    if pref:
        stg.columns = [pref+col_name for col_name in stg.columns]
    
    df = df.join(stg)
    
    return df
```

```python
# Expand categories from a string into separate columns

def expand_categories_to_cols(df, col_name, sep, new_cols):
    """
    expand_categories_to_cols creates columns-categories/subcategories out of categorical values of a given column
        The categories in the column may be represented by compound values, concatenated with a separator.
    For example "LC - Sites, Resources & Reference - Games & Simulations", in which case they can be split into three corresponding columns:
        "category_level1, category_level2, category_level3" with values in the corresponding columns: "LC", "Sites, Resources & Reference", "Games & Simulations"
    
    Parameters:
        df - Source DataFrame
        col_name - Column name in df, which is used for expansion
        sep - Symbol-separator between values
        new_cols - List of new column names
                
    Returns:
        A DataFrame same as original one with added categorical columns
                
    Example:
        # Load data
        >>>> df_district_info, df_products = load_reference_data()

        # In the example below we take Products data (represented as df_products DataFrame) and make the categories splitting from its 'Primary Essential Function' column.
        # The 'Primary Essential Function' will remain intact.
        >>>> df_p = expand_categories_to_cols(df_products, 'Primary Essential Function', sep=' - ', new_cols=["function_level1", "function_level2", "function_level3"])
    """

    new_df = pd.DataFrame(df[col_name].str.split(sep, len(new_cols)).tolist(), columns=new_cols)
    
    new_df = new_df.reset_index()
    df = df.reset_index()
    
    df = pd.concat([df, new_df], axis=1)
    
    return df
```

```python
# Cleaning Product Data

def clean_product_data(df):
    """
    clean_product_data makes Product data more consistent, like correcting 'Primary Essential Function' spelling.
        For example, 'Sites, Resources & References' replace to 'Sites, Resources & Reference' for some products to make it consistent with the rest of
        the products with similar category.
    
    Parameters:
        df - source DataFrame, loaded from districts_info.csv
        To load Products data, you can use load_reference_data function, like followng:
        >>>> df_district_info, df_products = load_reference_data()
        
    Returns:
        processed DataFrame
        
    Example:
        # Load data
        >>>> df_district_info, df_products = load_reference_data()
        
        # Pre-process Products data
        >>>> df_products = enhance_products_data(df_products)
    """
    
    df.loc[df["Primary Essential Function"] == "LC - Sites, Resources & References - Learning Materials & Supplies", "Primary Essential Function"] = "LC - Sites, Resources & Reference - Learning Materials & Supplies"
    
    return df
```

```python
# District Data pre-processing

def prepare_district_data(df):
    """
    prepare_district_data does all the necessary steps to pre-process District data
    
    Parameters:
        df - source DataFrame, loaded from districts_info.csv
        To load District data, you can use load_reference_data function, like followng:
        >>>> df_district_info, df_products = load_reference_data()
    
    Returns:
        DataFrame prepared with various transformations:
            1. Transforms range values (columns like 'pct_black/hispanic' etc.) into two columns with lower and upper end of the ranges
            
    Example:
        # Load data
        >>>> df_district_info, df_products = load_reference_data()
        
        # Prepare District data
        >>>> df_district_info = prepare_district_data(df_district_info)
    """
    
    df["county_connections_ratio"].fillna("[0.18, 1[", inplace=True)
    df.loc[df["county_connections_ratio"].str.contains("1, 2"), "county_connections_ratio"] = "[0.18, 1["
    
    df["pct_black/hispanic_lower"] = df["pct_black/hispanic"].apply(lambda b : get_lower_range(b))
    df["pct_black/hispanic_upper"] = df["pct_black/hispanic"].apply(lambda b : get_upper_range(b))
    cols = df_district_info.loc[:, "pct_black/hispanic_lower":"pct_black/hispanic_upper"]
    df["pct_black/hispanic_mean"] = cols.mean(axis=1)

    df["pct_free/reduced_lower"] = df["pct_free/reduced"].apply(lambda b : get_lower_range(b))
    df["pct_free/reduced_upper"] = df["pct_free/reduced"].apply(lambda b : get_upper_range(b))
    cols = df_district_info.loc[:, "pct_free/reduced_lower":"pct_free/reduced_upper"]
    df["pct_free/reduced_mean"] = cols.mean(axis=1)

    df["county_connections_ratio_lower"] = df["county_connections_ratio"].apply(lambda b : get_lower_range(b))
    df["county_connections_ratio_upper"] = df["county_connections_ratio"].apply(lambda b : get_upper_range(b))
    cols = df_district_info.loc[:, "county_connections_ratio_lower":"county_connections_ratio_upper"]
    df["county_connections_ratio_mean"] = cols.mean(axis=1)

    df["pp_total_raw_lower"] = df["pp_total_raw"].apply(lambda b : get_lower_range(b))
    df["pp_total_raw_upper"] = df["pp_total_raw"].apply(lambda b : get_upper_range(b))
    cols = df_district_info.loc[:, "pp_total_raw_lower":"pp_total_raw_upper"]
    df["pp_total_raw_mean"] = cols.mean(axis=1)
    
    df["pct_black/hispanic_mean"].fillna(df["pct_black/hispanic_mean"].median(), inplace=True)
    df["pct_free/reduced_mean"].fillna(df["pct_free/reduced_mean"].median(), inplace=True)
    df["county_connections_ratio_mean"].fillna(df["county_connections_ratio_mean"].median(), inplace=True)
    df["pp_total_raw_mean"].fillna(df["pp_total_raw_mean"].median(), inplace=True)
    
    return df
```

```python
# Products Data pre-processing

def prepare_products_data(df):
    """
    prepare_products_data does all the necessary steps to pre-process Products data
    
    Parameters:
        df - source DataFrame, loaded from products_info.csv
        To load Products data, you can use load_reference_data function, like followng:
        >>>> df_district_info, df_products = load_reference_data()
    
    Returns:
        DataFrame prepared with various transformations:
            1. Expand 'Sector(s)' column into one-hot categorical attributes
            2. Re-categorise some products
            3. Delete some general purpose products
            
    Example:
        # Load data
        >>>> df_district_info, df_products = load_reference_data()
        
        # Prepare District data
        >>>> df_products = prepare_products_data(df_products)
    """
    
    df = clean_product_data(df)
#     df = enhance_products_data(df)
    df = expand_categories_one_hot(df, 'Sector(s)', replace_str='; ', replace_with=';', pref='sec_')
    df = expand_categories_to_cols(df, 'Primary Essential Function', sep=' - ', new_cols=["function_level1", "function_level2", "function_level3"])

    return df
```

```python
# Engagement Data pre-processing

from datetime import datetime

def prepare_engagement_data(df):
    """
    prepare_engagement_data does all the necessary steps to pre-process Engagement data
    
    Parameters:
        df - source DataFrame, loaded from Engagement files. For the load you can use load_engagement_data function, as following
        >>>> df_engagement_data = load_engagement_data()
    
    Returns:
        DataFrame prepared with various transformations:
            1. Fill empty values of 'engagement_index' with zero
            2. Transform 'Time' column to datetime format
            
    Example:
        # Load Engagement data
        >>>> df_engagement_data = load_engagement_data()
        
        # Prepare Engagement data
        >>>> df_engagement_data = prepare_engagement_data(df_engagement_data)
    """
    
    df["engagement_index"].fillna(0, inplace=True)
    df.loc[:, "time"] = df["time"].apply(lambda b : datetime.fromisoformat(b))
    
    # Ignore data withing summer school holidays, as this time a natural drop in educational activity is happening
    index_list= df.time[(df.time >= "2020-06-01 00:00:00") & (df.time <= "2020-08-31 00:00:00")].index
    df.drop(index_list , inplace = True)
    
    return df
```

```python
# Merge data sets

def merge_data(d_engagement_data, d_products, d_district_info):
    """
    merge_data merges Engagement data, Products data, and District info data
    
    Parameters:
        d_engagement_data - Engagement Data
        d_products - Products Data
        d_district_info - District Info Data
        
        For the load you can use load_engagement_data function, as following
        >>>> df_engagement_data = load_engagement_data()
        >>>> df_district_info, df_products = load_reference_data()
    
    Returns:
        Merged DataFrame, containing columns from three data sets
            
    Example:
        # Load Engagement data
        >>>> df_engagement_data = load_engagement_data()
        
        # Prepare Engagement data
        >>>> df_engagement_data = prepare_engagement_data(df_engagement_data)
        
        # Merge the data sets
        >>>> df_merged = merge_data(df_engagement_data, df_products, df_district_info)
    """
    d1 = pd.merge(
        d_engagement_data,
        d_products,
        how="inner",
        on=None,
        left_on=["lp_id"],
        right_on=["LP ID"],
        left_index=False,
        right_index=False,
        sort=True,
        suffixes=("_x", "_y"),
        copy=True,
        indicator=False,
        validate=None,
    )
    
    d2 = pd.merge(
        d1,
        d_district_info,
        how="inner",
        on=None,
        left_on=["district_id"],
        right_on=["district_id"],
        left_index=False,
        right_index=False,
        sort=True,
        suffixes=("_x", "_y"),
        copy=True,
        indicator=False,
        validate=None,
    )
    
    return d2
```

```python
from sklearn.preprocessing import minmax_scale
import plotly.express as px

def get_annotation(p_text, p_x):
    annotation = dict(
            x=p_x,
            y="1",
            text=p_text,
            textangle=0,
            ax=0,
            ay=-35,
            showarrow=True,
            font=dict(
                color="black",
                size=12
            ),
            arrowcolor="black",
            arrowsize=0.3,
            arrowwidth=1,
            arrowhead=1)
    
    return annotation


def show_products_usage_trend(df, group_by):

    df_by_level = df[["time", group_by]]
    df_all_grouped_level = df_by_level.groupby([pd.Grouper(key="time", freq='W-MON'), group_by], as_index=True).size().reset_index(name="cnt")

    # Min-max scale to make it more clear on the graph
    df_all_grouped_level[["cnt"]] = df_all_grouped_level.groupby(group_by).cnt.transform(lambda x : minmax_scale(x))
    
    fig = px.scatter(df_all_grouped_level, x="time", y="cnt", color=group_by, trendline="lowess")
    fig.update_traces(visible=False, selector=dict(mode="markers"))
    fig.update_traces(showlegend=True)
    
    fig.update_layout(
    annotations=[
        get_annotation("K-12 public schools closed", "2020-03-17"),
        get_annotation("SAT", "2020-08-29"),
        get_annotation("SAT", "2020-09-26"),
        get_annotation("SAT", "2020-10-03"),
        get_annotation("SAT", "2020-11-20"),
        get_annotation("SAT", "2020-12-05")
    ]
    )
    
    fig.show()
```

<span style='font-size:20px'>&#x2605;</span><span style='font-family:"Times New Roman"; font-size:20px; margin-left:20px'>We will be analysing data based on average engagement index per day, grouped by states.</span>

```python
# Main functions for State Interventions analysis
# We analyse the how events affect engagement by comparing the slopes of linear regression of the engagement index before and after the event (slope ratio)

from scipy.stats import linregress
from datetime import datetime

# Save the slopes in a dictionary
#  slopes = {
#      "Attribute Name 1":[<slope ratio in state 1>, <slope ratio in state 2>, ...],
#      "Attribute Name 2":[<slope ratio in state 1>, <slope ratio in state 2>, ...],
#      ...
#         }
#  Then, from the dictionary we will select top attributes, for which standard deviation of slope ratios is the least

slopes = dict()
slopes_states = dict()
ling_list = []

def calculate_slope(df):
    x = np.arange(len(df))
    y = df["eng_idx_mean"]
    
    ret = linregress(x, y)
    
    ling_list.append({"x":x, "y":y, "ret":ret})
    
    return ret[0]

def add_slope(df, attribute, state, start_date, end_date):
    # Each run of this function adds a new result into list of slopes of the passing Attribute
    
    year_beginning = datetime(2020, 1, 1)
    year_end = datetime(2020, 12, 31)
    
    event_length = min((start_date - year_beginning), (end_date - start_date))
    start_before_event = start_date - event_length
    end_before_event = start_date
    end_after_event = start_date + event_length

    df_after_event = df[(df["state"]==state) & (df["time"]>=start_date) & (df["time"]<=end_after_event)]
    df_before_event = df[(df["state"]==state) & (df["time"]>=start_before_event) & (df["time"]<=end_before_event)]
    
    if len(df_after_event) == 0 or len(df_before_event) == 0:
        return
    
    slope_after_event = calculate_slope(df_after_event)
    slope_before_event = calculate_slope(df_before_event)
    
    if np.isnan(slope_after_event) or np.isnan(slope_before_event):
        return
    
    slopes_ratio = slope_after_event / slope_before_event
    
    if not np.isnan(slopes_ratio):
        if not attribute in slopes:
            slopes[attribute] = []
            slopes_states[attribute] = dict()
            
        slopes[attribute].append(slopes_ratio)
        slopes_states[attribute][state] = ({"state":state,
                                         "slopes_ratio":slopes_ratio,
                                         "slope_before_event":slope_before_event,
                                         "slope_after_event":slope_after_event,
                                         "start_before_event":start_before_event,
                                         "end_before_event":end_before_event,
                                         "start_date":start_date,
                                         "end_after_event":end_after_event,
                                         "event_length":event_length,
                                         "year_beginning":year_beginning,
                                         "end_date":end_date,
                                         "year_end":year_end,
                                         "event_length":event_length
                                        }
                                       )
```

```python
# What attributes are we testing?
# The attributes is a list of (<start date>, <end date>) of the event
# for example: ("STEMERG", "STEMERGEND") means that we are measuring how "State of emergency issued" affected engagement index

attributes_to_test = [
    ("STEMERG", "STEMERGEND"),
    ("CLSCHOOL", "END_OF_YEAR"),
    ("CLDAYCR", "OPNCLDCR"),
    ("STAYHOME", "END_STHM"),
    ("CLGYM", "ENDGYM"),
    ("CLMOVIE", "END_MOV"),
    ("EMSTART", "EMEND"),
    ("EMSTART2", "EMEND2"),
    ("INITIATIONBANSTART", "INITIATIONBANEND"),
    ("HEARINGBANSTART", "HEARINGBANEND"),
    ("HEARINGBANSTART2", "HEARINGBANEND2"),
    ("ENFORCEBANSTART", "ENFORCEBANEND"),
    ("C19START", "C19END"),
    ("PAYSTART", "PAYEND"),
    ("CARESSTART", "CARESEND"),
    ("CDCSTART", "CDCEND"),
    ("LATESTART", "LATEEND"),
    ("SMSTART", "SMEND"),
    ("SNAPALLO", "END_OF_YEAR"),
    ("SNAPEBT20", "END_OF_YEAR"),
    ("SNAPSUSP", "END_OF_YEAR"),
    ("MED1135W", "END_OF_YEAR"),
    ("ACAENROL", "END_OF_YEAR"),
    ("TLHLAUD", "END_OF_YEAR"),
    ("TLHLMED", "END_OF_YEAR"),
    ("VISITPER", "VISITRES"),
    ("NOVISIT2", "VISITRES2"),
    ("ELECPRCR", "ENDELECP"),
    ("WV_WTPRD", "END_OF_YEAR"),
    ("EBSTART", "EBEND"),
    ("20EBSTART", "20EBEND")
]
```

```python
# Prepare data for analysis

df_by_day = df_merged[["time", "state", "engagement_index"]].dropna()
df_grouped_eng_idx = df_by_day.groupby(["time", "state"]).engagement_index.mean().reset_index(name="eng_idx_mean")
```

```python
# Load Policies Data
df_policies = load_policies()
df_policies_prep = policies_set_dates(df_policies)
df_policies_prep["END_OF_YEAR"] = datetime(2020, 12, 31)
```

```python
# Calculate how events affected engagements index

from datetime import timedelta

slopes = dict()
slopes_states = dict()

for idx, policy_record in df_policies_prep.iterrows():
    for attributes in attributes_to_test:
        end_date = policy_record[attributes[0]] + timedelta(days=30)
        add_slope(df_grouped_eng_idx, attributes[0], policy_record["STATE"], policy_record[attributes[0]], end_date)
```

```python
# Compute standard deviation for each attribute and sort so on top we have the most influencing attributes

attributes = list()

for attribute in slopes.items():
    # consider only events hapened in 5 or more states
    if len(attribute[1]) >= 7:
        # remove outliers
        slopes[attribute[0]].remove(max(slopes[attribute[0]]))
        slopes[attribute[0]].remove(min(slopes[attribute[0]]))
        
        # consider only pivoting events, after which the trends went the opposite direction
        # that is, the slopes ratio must be negative
        trends = [e for e in attribute[1] if e < 0]

        std = np.std(trends)
        mean = np.mean(trends)
        
        if not np.isnan(std):
            attributes.append({"attr":attribute[0], "std":std, "mean":mean})
        
attributes_top = sorted(attributes, key=lambda k: k["mean"], reverse=False)
```
