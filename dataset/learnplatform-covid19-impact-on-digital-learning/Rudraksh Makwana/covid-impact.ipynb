```python
prod_info = pd.read_csv('/kaggle/input/learnplatform-covid19-impact-on-digital-learning/products_info.csv')
prod_info.head()
```

```python
prod_info.rename(columns={'LP ID': 'lp_id'}, inplace=True)
```

```python
dist_info = pd.read_csv('/kaggle/input/learnplatform-covid19-impact-on-digital-learning/districts_info.csv')
dist_info.head()
```

```python
path = '/kaggle/input/learnplatform-covid19-impact-on-digital-learning/engagement_data'

frame = []

for district in dist_info.district_id.unique():
    df = pd.read_csv(f'{path}/{district}.csv', index_col=None, header=0)
    df["district_id"] = district
    frame.append(df)
    
    
engagement = pd.concat(frame)
engagement = engagement.reset_index(drop=True)
```

```python
inter_data = engagement.join(dist_info.set_index(['district_id']), on = 'district_id')
data = inter_data.join(prod_info.set_index(['lp_id']), on='lp_id')
data
```
