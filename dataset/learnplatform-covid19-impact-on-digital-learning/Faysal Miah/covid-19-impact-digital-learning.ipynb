```python
# Importing the libreries
import os, glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns 
sns.set(style="white")

%matplotlib inline
```

```python
# Importing the datasets

# Engagment data
eng_path = "../input/learnplatform-covid19-impact-on-digital-learning/engagement_data"
files = glob.glob(eng_path+"/*.csv")

all_files = []
for file in files:
    df = pd.read_csv(file, index_col=None, header=0)
    district_id = file.split("/")[4].split(".")[0]
    df["district_id"] = district_id
    all_files.append(df)
    
engagment_data = pd.concat(all_files)
engagment_data = engagment_data.reset_index(drop=True)


# Districts info data
districts_info = pd.read_csv("../input/learnplatform-covid19-impact-on-digital-learning/districts_info.csv")
# Products info data
products_info = pd.read_csv("../input/learnplatform-covid19-impact-on-digital-learning/products_info.csv")
```

```python
engagment_data.rename(columns={"lp_id": "LP ID"}, inplace=True)
merged=pd.merge(engagment_data, products_info, on= "LP ID")
m=merged.groupby("Product Name")["pct_access"].mean().sort_values(ascending=False).head(10)

engagment_data.rename(columns={"lp_id": "LP ID"}, inplace=True)
merged=pd.merge(engagment_data, products_info, on= "LP ID")
n=merged.groupby("Product Name")["engagement_index"].sum().sort_values(ascending=False).head(10)

# plot
plt.figure(figsize=(15,4))

plt.subplot(121)
plt.bar(m.index, m.values, color=["#6930c3","#5e60ce","#0096c7","#48cae4","#ade8f4","#ff7f51","#ff9b54","#ffbf69"])
plt.xlabel('Product Name')
plt.xticks(rotation=90)
plt.ylabel('Mean percentage of students')
plt.title("With atleast one-page load event")

plt.subplot(122)
plt.bar(n.index, n.values, color=["#4f000b","#720026","#ce4257","#ff7f51","#ff9b54"])
plt.xlabel('Product Name')
plt.xticks(rotation=90)
plt.ylabel('Page-load per 1000 students')
plt.title("With number of page-load per 1000 students")
```
