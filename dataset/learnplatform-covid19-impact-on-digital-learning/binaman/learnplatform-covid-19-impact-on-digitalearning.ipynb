```python
def displayMissingValue(df):
    print(".......................")
    print("")
    print(df.isna().sum())
    print("")
    totalCells = np.product(df.shape)
    missingCount = df.isnull().sum()
    totalMissing = missingCount.sum()
    print("The dataset contains", round(((totalMissing / totalCells) * 100), 2), "%", "missing values.")
    print("")
    msno.bar(df,color='#7209b7', sort="ascending", figsize=(10,5), fontsize=12)
    plt.show()
    print(".......................")
```

```python
def plot_count(df:pd.DataFrame, column:str,title:str) -> None:
    plt.figure(figsize=(12, 7))
    sns.countplot(data=df, x=column) 
    plt.xticks(rotation=75, fontsize=14)
    plt.title(title,font="Serif", size=20)
    plt.show()
def group_donut(grouped_data,title: str):
    grouped_data.plot.pie(subplots=True,figsize=(18, 9),autopct="%.1f%%",pctdistance=0.85)
    # add a circle at the center to transform it in a donut chart
    my_circle=plt.Circle( (0,0), 0.7, color='white')
    p=plt.gcf()
    p.gca().add_artist(my_circle)
    plt.title(title,font="Serif", size=20)
    plt.show()
def bar_p(df:pd.DataFrame, column:str,title:str):
    plt.figure(figsize=(18, 9))
    sns.countplot(y=column, data=prod_df, order=df[column].value_counts().head(10).index,color = "#a265b8")
    plt.title(title,font="Serif", size=20)
    plt.show()
```
