Observation : Just saw the data. The districts data and one file from the engegement data has been seen.The data seems to be highly polluted. i:e: it has so much null values and outliers. We can start with merging all the engagement data to a single dataframe.

```python
len(os.listdir('../input/learnplatform-covid19-impact-on-digital-learning/engagement_data'))
```

```python
import glob

files = glob.glob('../input/learnplatform-covid19-impact-on-digital-learning/engagement_data/*.csv')
all_districts = None

for i, f in enumerate (files):
    if i == 0:
        all_districts = pd.read_csv(f)
        all_districts['district_id'] = os.path.basename(f)[:4]
    else:
        tmp = pd.read_csv(f)
        tmp['district_id'] = os.path.basename(f)[:4]
        all_districts = all_districts.append(tmp)
```

```python
all_districts.shape
```

```python
all_districts.head()
```

```python
all_districts.isna().sum()
```

```python
all_districts.dropna(how='any',inplace=True)
```
