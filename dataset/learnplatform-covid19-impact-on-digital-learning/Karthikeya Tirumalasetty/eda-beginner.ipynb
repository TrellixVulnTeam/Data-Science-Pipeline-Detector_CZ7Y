```python
districts_df = pd.read_csv("../input/learnplatform-covid19-impact-on-digital-learning/districts_info.csv")
districts_df.head()
```

```python
print(districts_df.shape)
print()
print(districts_df.info())
print("+"*30)
districts_df.isna().sum()#to find nan values present
```

```python
path = '../input/learnplatform-covid19-impact-on-digital-learning/engagement_data' 
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    district_id = filename.split("/")[4].split(".")[0]
    df["district_id"] = district_id
    li.append(df)
    
engage_df = pd.concat(li)
engage_df = engage_df.reset_index(drop=True)
engage_df.head()
```

```python
print(engage_df.info())
print("+"*30)
engage_df.isna().sum()
```

```python
fig, axs = plt.subplots(3, figsize=(15,15))

sns.barplot(ax=axs[0], x=products_df.columns, y=products_df.isnull().sum())
axs[0].set_title('Missing Values for Products',fontsize=14)

sns.barplot(ax=axs[1], x=districts_df.columns, y=districts_df.isnull().sum())
axs[1].set_title('Missing Values for Districts',fontsize=14)

sns.barplot(ax=axs[2], x=engage_df.columns, y=engage_df.isnull().sum())
axs[2].set_title('Missing Values for Engagements',fontsize=14)

plt.show()
```
