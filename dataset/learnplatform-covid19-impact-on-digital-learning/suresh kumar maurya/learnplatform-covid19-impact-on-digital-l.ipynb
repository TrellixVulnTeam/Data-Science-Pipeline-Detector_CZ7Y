```python
missing_values=("NaN",np.nan)
products=pd.read_csv("../input/learnplatform-covid19-impact-on-digital-learning/products_info.csv",na_values=missing_values)
```

```python
products.isnull().sum()
```

```python
products.isnull().any()
```

```python
products.dropna(how="all")
```

```python
null_var=products.isnull().sum()/products.shape[0]*100
```

```python
drop_columns= null_var[null_var > 4].keys()
```
