#Understand a column with plot(df, x)

After getting an overview of the dataset, we can thoroughly investigate a column of interest x using plot(df, x). The output is of plot(df, x) is different for numerical and categorical columns.

When x is a categorical column, it computes column statistics, and plots a bar chart and pie chart.

It Includes WordCloud! Plus Word frequency and length.

```python
plot(df, "Product Name")
```

When x is a numerical column (below), it computes column statistics, and generates a histogram, kde plot, box plot and qq-normal plot:

```python
plot(df, "LP ID")
```

When x and y are both categorical columns, it plots a nested bar chart, stacked bar chart and heat map.

When x and y are both numerical columns, it generates a scatter plot, hexbin plot and box plot. Though we don't have two numerical columns here.

```python
plot(df, "Provider/Company Name","Primary Essential Function")
```

```python
import holoviews as hv
from holoviews import opts
hv.extension('bokeh')
```

```python
#Code by Kohei-mu https://www.kaggle.com/koheimuramatsu/industrial-accident-causal-analysis/notebook

sector_cnt = np.round(df['Sector(s)'].value_counts(normalize=True) * 100)
hv.Bars(sector_cnt).opts(title="Sector Count", color="green", xlabel="Sector", ylabel="Percentage", yformatter='%d%%')\
                .opts(opts.Bars(width=500, height=300,tools=['hover'],show_grid=True)).opts(xrotation=45)
```

```python
#Code by Kohei-mu https://www.kaggle.com/koheimuramatsu/industrial-accident-causal-analysis/notebook

loc_cnt = np.round(df1['locale'].value_counts(normalize=True) * 100)
hv.Bars(loc_cnt).opts(title="Locale Count", color="yellow", xlabel="Locale", ylabel="Percentage", yformatter='%d%%')\
                .opts(opts.Bars(width=500, height=300,tools=['hover'],show_grid=True)).opts(xrotation=45)
```

```python
#Code by Kohei-mu https://www.kaggle.com/koheimuramatsu/industrial-accident-causal-analysis/notebook

#Fixed by Des desalegngeb https://www.kaggle.com/desalegngeb
#https://www.kaggle.com/desalegngeb/bokeh-visualization-library-guide-for-beginners

prim_cnt = np.round((df['Primary Essential Function'].value_counts(normalize=True) *100).head()) 
hv.Bars(prim_cnt).opts(title="1ary Essential Function Count", color="red", xlabel="1ary Essential function", ylabel="Percentage", yformatter='%d%%')\
                .opts(opts.Bars(width=500, height=300,tools=['hover'],show_grid=True)).opts(xrotation=45)
```

```python
#Code by Kohei-mu https://www.kaggle.com/koheimuramatsu/industrial-accident-causal-analysis/notebook

#Fixed by Des https://www.kaggle.com/desalegngeb/bokeh-visualization-library-guide-for-beginners

sta_cnt = np.round((df1['state'].value_counts(normalize=True) *100).head())
hv.Bars(sta_cnt[::-1]).opts(title="State Count", color="cyan", xlabel="US States", ylabel="Percentage", xformatter='%d%%')\
                .opts(opts.Bars(width=600, height=600,tools=['hover'],show_grid=True,invert_axes=True))
```

```python
#Code by Kohei-mu https://www.kaggle.com/koheimuramatsu/industrial-accident-causal-analysis/notebook

#Fixed by Des https://www.kaggle.com/desalegngeb/bokeh-visualization-library-guide-for-beginners

prov_cnt = np.round((df['Provider/Company Name'].value_counts(normalize=True) *100).head())
hv.Bars(prov_cnt[::-1]).opts(title="Provider/Company Count", color="purple", xlabel="Providers/Companys", ylabel="Percentage", xformatter='%d%%')\
                .opts(opts.Bars(width=600, height=600,tools=['hover'],show_grid=True,invert_axes=True))
```

```python
#Code by Kohei-mu https://www.kaggle.com/koheimuramatsu/industrial-accident-causal-analysis/notebook

#Here the entries have only one count each and the bar graphs will appear flat. I kept [:10].

url_cnt = np.round(df['URL'].value_counts(normalize=True) [:10])
hv.Bars(url_cnt[::-1]).opts(title="URL Count", color="cyan", xlabel="10 Url", ylabel="Percentage", xformatter='%d%%')\
                .opts(opts.Bars(width=600, height=600,tools=['hover'],show_grid=True,invert_axes=True))
```
