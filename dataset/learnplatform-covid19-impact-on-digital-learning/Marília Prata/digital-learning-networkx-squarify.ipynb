The squarified tree map algorithm, was invented by Mark Bruls, Kees Huizing, and Jarke J. van Wijk in a 2000 paper as an improvement on the “classic” tree map, which was in turn invented by Ben Shneiderman in a pioneering 1992 paper. This algorithm is used to lay out probably all the pleasing tree maps you see around.

```python
df['locale'].value_counts()
```

```python
df['locale'] = 'locale'
locale = df.truncate(before = -1, after = 15)
```

```python
import networkx as nx

locale = nx.from_pandas_edgelist(locale, source = 'locale', target = "state", edge_attr = True)
```

```python
#Code by Roshan Sharma https://www.kaggle.com/roshansharma/market-basket-analysis

import warnings
warnings.filterwarnings('ignore')

plt.rcParams['figure.figsize'] = (20, 20)
pos = nx.spring_layout(locale)
color = plt.cm.Wistia(np.linspace(0, 15, 1))
nx.draw_networkx_nodes(locale, pos, node_size = 15000, node_color = 'r')
nx.draw_networkx_edges(locale, pos, width = 3, alpha = 0.6, edge_color = 'blue')
nx.draw_networkx_labels(locale, pos, font_size = 20, font_family = 'sans-serif')
plt.axis('off')
plt.grid()
plt.title('Closure during Pandemic', fontsize = 40)
plt.show()
```

```python
y = df["state"].value_counts().head().to_frame()
y.index
y.head()
```

```python
#Code by Roshan Sharma https://www.kaggle.com/roshansharma/market-basket-analysis

import squarify
# plotting a tree map

plt.rcParams['figure.figsize'] = (16, 16)
color = plt.cm.cool(np.linspace(0, 1,10))
squarify.plot(sizes = y.values, label = y.index, alpha=.8, color = color)
plt.title('Tree Map of States')
plt.axis('off')
plt.show()
```

```python
df['states'] = 'State'
states = df.truncate(before = -1, after = 15)

df.head()
```

```python
import networkx as nx

states = nx.from_pandas_edgelist(states, source = 'states', target = 'locale', edge_attr = True)
```

```python
#Code by Roshan Sharma https://www.kaggle.com/roshansharma/market-basket-analysis


plt.rcParams['figure.figsize'] = (20, 20)
pos = nx.spring_layout(states)
color = plt.cm.Wistia(np.linspace(0, 15, 1))
nx.draw_networkx_nodes(states, pos, node_size = 15000, node_color = color)
nx.draw_networkx_edges(states, pos, width = 3, alpha = 0.6, edge_color = 'black')
nx.draw_networkx_labels(states, pos, font_size = 20, font_family = 'sans-serif')
plt.axis('off')
plt.grid()
plt.title('Locales', fontsize = 40)
plt.show()
```
