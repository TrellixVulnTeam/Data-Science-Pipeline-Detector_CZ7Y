```python
#Code by Bojan Tunguz https://www.kaggle.com/tunguz/tps-apr-21-tsne-and-umap-embeddings-with-rapids/notebook

import cudf, cuml
import cupy as cp

from cuml.manifold import TSNE, UMAP
import matplotlib.pyplot as plt
from matplotlib.pyplot import ylim, xlim
%matplotlib inline
```

```python
df = pd.read_csv("/kaggle/input/learnplatform-covid19-impact-on-digital-learning/engagement_data/5802.csv", low_memory=False)
print(df.shape)
df.head()
```

```python
df_grp = df.groupby(["lp_id"])[["pct_access","engagement_index"]].sum().reset_index()
df_grp.head()
```

```python
columns = df_grp.columns
columns
```

```python
df_grp[columns].head()
```

```python
%%time
tsne = TSNE(n_components=2)
df_grp_2D = tsne.fit_transform(df_grp[columns].values)
```

```python
#Code by Bojan Tunguz https://www.kaggle.com/tunguz/tps-apr-21-tsne-and-umap-embeddings-with-rapids/notebook

plt.scatter(df_grp_2D[:,0], df_grp_2D[:,1], c=df_grp['lp_id'].values, s = 0.8);
```

```python
plt.scatter(df_grp_2D[:,0], df_grp_2D[:,1], c=df_grp['engagement_index'].values, s = 0.8);
```

```python
df1 = pd.read_csv("/kaggle/input/learnplatform-covid19-impact-on-digital-learning/districts_info.csv", low_memory=False)
print(df1.shape)
df1.head()
```

#Let's work with less columns to facilitate (myself)

```python
df1_grp = df1.groupby(["district_id"])[["state","locale", "pct_black/hispanic"]].sum().reset_index()
df1_grp.head()
```

```python
df1_grp = pd.get_dummies(df1_grp, drop_first=True)
```

```python
df1_grp.head()
```

```python
columns = df1_grp.columns
columns
```

```python
df1_grp[columns].head()
```

```python
%%time
tsne = TSNE(n_components=2)
df1_grp_2D = tsne.fit_transform(df1_grp[columns].values)
```

```python
plt.scatter(df1_grp_2D[:,0], df1_grp_2D[:,1], c=df1_grp['district_id'].values, s = 0.8);
```

```python
plt.scatter(df1_grp_2D[:,0], df1_grp_2D[:,1], c=df1_grp['pct_black/hispanic_[0.2, 0.4['].values, s = 0.8);
```
