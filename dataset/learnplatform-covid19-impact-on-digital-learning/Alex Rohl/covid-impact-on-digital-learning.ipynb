```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import glob
```

```python
path = '../input/learnplatform-covid19-impact-on-digital-learning/engagement_data' 
all_files = glob.glob(path+'/../*md')
print(open(all_files[0]).read())
```

```python
path = '../input/learnplatform-covid19-impact-on-digital-learning/engagement_data' 
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    district_id = filename.split("/")[4].split(".")[0]
    df["district_id"] = district_id
    li.append(df)
    
engagement_df = pd.concat(li)
engagement_df = engagement_df.reset_index(drop=True)
engagement_df.head()
```

```python
district_id = 2956
fig = plt.figure(figsize=(20,10))
fig.suptitle(f"Engagement Index for {districts_df[districts_df['district_id']==district_id]['state']}")
sns.lineplot(data=engagement_df[(engagement_df['district_id']==str(district_id)) & (engagement_df['lp_id']==13117)],x='time',y='engagement_index',hue='lp_id')
fig.autofmt_xdate()
```

```python
engagement_df['district_id'] = engagement_df['district_id'].astype(int)
merged = pd.merge(engagement_df, districts_df, on = 'district_id')
products_df['lp_id'] = products_df['LP ID'].astype(int)
merged = pd.merge(merged, products_df, on = 'lp_id')
merged.head()
```

```python
group_by_primary = merged.groupby(['Primary Essential Function', 'state','time']).mean().reset_index()
```

```python
common_functions = merged['Primary Essential Function'].value_counts()[:10].index
```

```python
fig = plt.figure(figsize=(20,10))
sns.scatterplot(data = group_by_primary[group_by_primary['Primary Essential Function'].isin(common_functions)], x='time', y='engagement_index', hue = 'Primary Essential Function', alpha=0.2)
```

```python
pct_freelunch_df = merged.groupby(['pct_free/reduced', 'state','time','Sector(s)']).mean().reset_index()
```

```python
pct_pp_total_df = merged.groupby(['pp_total_raw', 'state','time','Sector(s)']).mean().reset_index()   
```
