```python
import folium
from folium.plugins import HeatMap
from folium.plugins import FastMarkerCluster
import glob
from geopy.geocoders import Nominatim
```

```python

map_ = folium.Map(location=[40, -80], tiles='Stamen Terrain', zoom_start=4.2,attr='default')
map_
```

```python
locations = district_df['state'].unique()
geolocator=Nominatim(user_agent="app")
lat=[]
lon=[]
for location in locations:
    location = geolocator.geocode(location)    
    if location is None:
        lat.append(np.nan)
        lon.append(np.nan)
    else:
        lat.append(location.latitude)
        lon.append(location.longitude)
```

```python
count=district_df["state"].value_counts().values
geograhy_data=pd.DataFrame(list(zip(list(locations),lat,lon,count)),columns=["Location","Latitude","Longitude","Count"]).dropna()
geograhy_data
```

```python
FastMarkerCluster(geograhy_data[["Latitude","Longitude","Count"]],zoom=20,radius=20).add_to(map_)
HeatMap(geograhy_data[["Latitude","Longitude","Count"]],zoom=20,radius=20).add_to(map_)
map_
```

```python
black_precentage=[]
hispanic_precentage=[]
free_lunch=[]
reducrd_price_lunch=[]
total_expenditure_raw=[]
per_pupil_expenditure=[]

for i in district_df["pct_black/hispanic"]:
    b=i.split(",")
    black_precentage.append((b[0][1:]))
    hispanic_precentage.append((b[1][1:-1]))
    
for i in district_df["pct_free/reduced"]:
    b=i.split(",")
    free_lunch.append((b[0][1:]))
    reducrd_price_lunch.append((b[1][1:-1]))

for i in district_df["pp_total_raw"]:
    b=i.split(",")
    total_expenditure_raw.append((b[0][1:]))
    per_pupil_expenditure.append((b[1][1:-1]))
    


district_df_new=district_df.copy()
district_df_new["black_precentage"]=black_precentage
district_df_new["hispanic_precentage"]=hispanic_precentage
district_df_new["free_lunch"]=free_lunch
district_df_new["reducrd_price_lunch"]=reducrd_price_lunch
district_df_new["total_expenditure_raw"]=total_expenditure_raw
district_df_new["per_pupil_expenditure"]=per_pupil_expenditure


district_df_new.drop(["county_connections_ratio","pp_total_raw","pct_free/reduced","pct_black/hispanic"], axis=1, inplace=True)
district_df_new.head(5)
```

```python
district_df_new.describe()
```

```python
col_names=list(district_df_new.columns)[3:]
for name in col_names:
    ax = district_df_new[name].value_counts().sort_values()
    value = ax.values
    label = ax.index

    plt.figure(figsize=(14,8))
    plt.barh(y=label, width=value, edgecolor="k")
    for i in range(len(label)):
        plt.text(x=0.4,y=i-0.1,s=value[i],color="w",fontsize=12)
    plt.show()
```
