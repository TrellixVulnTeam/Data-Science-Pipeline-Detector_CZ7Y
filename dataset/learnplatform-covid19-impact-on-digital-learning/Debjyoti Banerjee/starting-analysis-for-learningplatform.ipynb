```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import re
import seaborn as sns
import plotly.express as px
```

```python
os.chdir('/kaggle/input/learnplatform-covid19-impact-on-digital-learning/engagement_data')
all_csv = os.listdir()
print(len(all_csv))
```

```python
dist_id = [i.split('.')[0] for i in all_csv]
```

```python
comb = []
for i in all_csv:
    df = pd.read_csv(i)
    df['district_id'] = i.split('.')[0]
    comb.append(df)
comb_engage_df = pd.concat(comb)
comb_engage_df = comb_engage_df.reset_index(drop=True)
comb_engage_df.head()
```

```python
comb_engage_df.shape
```

```python
comb_engage_df.info()
```

What % of the total is missing for each column.

```python
comb_engage_df.isnull().sum()/comb_engage_df.shape[0]
```

So 24% of the combined engagement_index is missing, which is a little big.

```python
all_rows = []
for i in all_csv:
    df = pd.read_csv(i)
    all_rows.append(df.shape[0])
print(all_rows)
```

```python
max_eng_index = []
for i in all_csv:
    df = pd.read_csv(i)
    max_eng_index.append(max(df['engagement_index']))
print(max_eng_index)
```

```python
df = pd.DataFrame({'Dist_Id': dist_id, 'Total data in a school district': all_rows,
                  'Engagement Index': max_eng_index})
df.head()
```

Now let's plot top 10 district ID which has more engagement.

```python
plt.figure(figsize = (12,6))
sns.set_style('darkgrid')
sns.barplot(data=df, x="Dist_Id", y="Total data in a school district", order = df.sort_values('Total data in a school district',ascending = False).Dist_Id[1:10], palette ='rocket')
```

**Now let's plot top 10 maximum engagement index of all the district schools.**

```python
plt.figure(figsize = (12,6))
sns.set_style('darkgrid')
sns.barplot(data=df, x="Dist_Id", y="Engagement Index", order = df.sort_values('Engagement Index',ascending = False).Dist_Id[1:10], palette ='viridis')
```

```python
comb_engage_df['time'] = pd.to_datetime(comb_engage_df['time'])
```

```python
comb_engage_df['Which Day'] = comb_engage_df['time'].dt.day_name()
comb_engage_df['Year'] = comb_engage_df['time'].dt.year
comb_engage_df['Month'] = (comb_engage_df['time'].dt.month_name())
```

**Now let's see as per the combined data, which month has more page load events.**

```python
sns.set(font_scale=1.2)
plt.figure(figsize = (12,6))
comb_engage_df['Month'].value_counts().plot(kind = 'bar')
plt.title('Frequecy showing which month has more data', size = 22)
```

Similarly let's see which day of the week has more page load events happened.

```python
sns.set(font_scale=1.2)
plt.figure(figsize = (12,6))
comb_engage_df['Which Day'].value_counts().plot(kind = 'bar', color = 'mediumvioletred')
plt.title('Frequecy showing which day of the week has more page load', size = 20)
plt.ylabel('Values are in range 10^6')
```
