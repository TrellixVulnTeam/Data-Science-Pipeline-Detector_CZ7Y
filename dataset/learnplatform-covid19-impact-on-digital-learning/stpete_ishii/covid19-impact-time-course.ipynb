```python
import numpy as np
import pandas as pd
import os
from tqdm.notebook import tqdm
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
paths=[]
for dirname, _, filenames in os.walk('../input/learnplatform-covid19-impact-on-digital-learning/engagement_data'):
    for filename in filenames:
        paths+=[os.path.join(dirname,filename)]
```

```python
data=pd.DataFrame()
for item in tqdm(paths):
    datai=pd.read_csv(item)
    datai['district_id']=int(item[:-4].split('/')[-1])
    data=pd.concat([data,datai])
data
```

```python
data2=data[['time','pct_access','engagement_index']]
data3=data2.groupby('time',as_index=False).sum()
data3
```

```python
fig=make_subplots(specs=[[{"secondary_y":False}]])
fig.add_trace(go.Scatter(x=data3['time'],y=data3['pct_access'],name='pct_access daily'),secondary_y=False,)
fig.add_trace(go.Scatter(x=data3['time'],y=data3['pct_access'].rolling(window=7).max(),name='pct_access 7-days max'),secondary_y=False,)
fig.update_layout(autosize=False,width=700,height=500,title_text="pct_access")
fig.update_xaxes(title_text="Date")
fig.update_yaxes(title_text="Pct Access",secondary_y=False)
fig.show()
```

```python
fig=make_subplots(specs=[[{"secondary_y":False}]])
fig.add_trace(go.Scatter(x=data3['time'],y=data3['engagement_index'],name='engagement_index daily'),secondary_y=False,)
fig.add_trace(go.Scatter(x=data3['time'],y=data3['engagement_index'].rolling(window=7).max(),name='engagement_index 7-days max'),secondary_y=False,)
fig.update_layout(autosize=False,width=700,height=500,title_text="engagement_index")
fig.update_xaxes(title_text="Date")
fig.update_yaxes(title_text="Engagement Index",secondary_y=False)
fig.show()
```
