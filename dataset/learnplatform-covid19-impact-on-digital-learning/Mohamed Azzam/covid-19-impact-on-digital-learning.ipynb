```python
def analysis_with_locale(locale, groupby = 'Sector(s)'):
    district_ids = districts[districts['locale'] == locale]['district_id']


    # create empty dataframe for engagement_data
    df_district = pd.DataFrame(columns=['time', 'lp_id', 'pct_access', 'engagement_index'])

    # read data with specific
    for i,id in enumerate(district_ids):
        engagement_data = pd.read_csv(f'../input/learnplatform-covid19-impact-on-digital-learning/engagement_data/{id}.csv')
        df_district = pd.concat([engagement_data,df_district])


    # create month column
    df_district['month'] = df_district['time'].astype('M8[D]').dt.month

    # merge data with product
    x_merge = pd.merge(df_district,products[['LP ID', 'Provider/Company Name', 
                     'Sector(s)', 'Primary Essential Function abbr']],
         left_on='lp_id', right_on='LP ID', how='left')


    x = x_merge.groupby(['lp_id', 'month', groupby]).mean()
    x.reset_index(level=[0,1,2], inplace=True)

    more_then_75_per = x['pct_access'].describe()['75%']
    x = x[x['pct_access'] >= more_then_75_per]

    return x    
```

```python
def analysis_with_state(state, groupby = 'Sector(s)'):
    district_ids = districts[districts['state'] == state]['district_id']


    # create empty dataframe for engagement_data
    df_district = pd.DataFrame(columns=['time', 'lp_id', 'pct_access', 'engagement_index'])

    # read data with specific
    for i,id in enumerate(district_ids):
        engagement_data = pd.read_csv(f'../input/learnplatform-covid19-impact-on-digital-learning/engagement_data/{id}.csv')
        df_district = pd.concat([engagement_data,df_district])


    # create month column
    df_district['month'] = df_district['time'].astype('M8[D]').dt.month

    # merge data with product
    x_merge = pd.merge(df_district,products[['LP ID', 'Provider/Company Name', 
                     'Sector(s)', 'Primary Essential Function abbr']],
         left_on='lp_id', right_on='LP ID', how='left')


    x = x_merge.groupby(['lp_id', 'month', groupby]).mean()
    x.reset_index(level=[0,1,2], inplace=True)

    more_then_75_per = x['pct_access'].describe()['75%']
    x = x[x['pct_access'] >= more_then_75_per]

    return x    
```
