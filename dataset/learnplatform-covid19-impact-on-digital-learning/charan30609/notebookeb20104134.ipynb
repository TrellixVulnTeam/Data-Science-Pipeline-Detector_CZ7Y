```python
products.columns = products.columns
list(products)
```

```python
feature_name= []
null_count= []
for i in districts.columns:
    feature_name.append(i)
    null_count.append(districts[i].isnull().sum())
print(feature_name)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,2,1])
ax.bar(feature_name,null_count)
plt.title("Bar Chart that show how many null values each feature contains ",
         fontdict={'fontsize': 18, 'fontweight': 'bold', 'color':'#138FBD'})
plt.show()
```

```python
districts.groupby(['state','locale'])['state'].count().plot.barh(figsize=(15,15), color='#55FF06')
plt.title("Bar chart shows the frequance of every Locale in each State ",
         fontdict={'fontsize': 18, 'fontweight': 'bold', 'color':'#581845'})
```

```python
districts.groupby(['state','locale'])['pp_total_raw'].value_counts().sort_index().plot.bar(figsize=(20,8), color="#FF8606")
plt.title("Bar chart that show the count of the school expenditure for each locale of each state",
         fontdict={'fontsize': 18, 'fontweight': 'bold', 'color':'#FF5733'})
```

```python
districts.groupby(['state','locale'])['pct_black/hispanic'].value_counts().sort_index().plot.bar(figsize=(25,8))
plt.title("Bar chart that show the count of the ratio of black students in each locale of each state",
         fontdict={'fontsize': 18, 'fontweight': 'bold', 'color':'blue'})
```

```python
df = products['Primary Essential Function'].dropna().reset_index(drop=True)
categories = df.str.split('-')
```

```python
category = []
subcategory = []
for i in range(len(categories)):
    category.append(categories[i][0])
    
    for cat in categories[i][1:]:
        subcategory.append(cat)
```

```python
title = 'Count of Subcategory'
fig = px.histogram(y=subcategory, title=title, labels={'y': 'Subcategory'}).update_yaxes(categoryorder='total ascending')
fig.show()
```
