## Read data - Engagement Data


```python
engagement_data_New_York_City = pd.read_csv('../input/learnplatform-covid19-impact-on-digital-learning/engagement_data/9536.csv')

engagement_data_New_York_Rural = pd.concat([pd.read_csv('../input/learnplatform-covid19-impact-on-digital-learning/engagement_data/2257.csv'), 
                                            pd.read_csv('../input/learnplatform-covid19-impact-on-digital-learning/engagement_data/7798.csv'), 
                                            pd.read_csv('../input/learnplatform-covid19-impact-on-digital-learning/engagement_data/8520.csv'),
                                            pd.read_csv('../input/learnplatform-covid19-impact-on-digital-learning/engagement_data/1470.csv'),
                                            pd.read_csv('../input/learnplatform-covid19-impact-on-digital-learning/engagement_data/9515.csv'),
                                           ])

engagement_data_New_York_Suburb = pd.concat([pd.read_csv('../input/learnplatform-covid19-impact-on-digital-learning/engagement_data/2940.csv'), 
                                            pd.read_csv('../input/learnplatform-covid19-impact-on-digital-learning/engagement_data/4775.csv'),                                             
                                           ])
```

# Daily Engagement 

```python
engagement_data_New_York_City['locale'] = 'city'
engagement_data_New_York_Rural['locale'] = 'rural'
engagement_data_New_York_Suburb['locale'] = 'suburb'

engagement_data_New_York = pd.concat([engagement_data_New_York_City,
                                      engagement_data_New_York_Rural,
                                      engagement_data_New_York_Suburb])
```

```python
engagement_data = pd.DataFrame(engagement_data_New_York)
engagement_data['time'] = engagement_data['time'].astype('datetime64')


# Merge engagement_data to product's info
engagement_data = engagement_data.merge(products_data, left_on='lp_id', right_on='ID',how='left')
engagement_data = engagement_data.drop(columns=['ID'])
engagement_data = engagement_data.set_index(['time'])
engagement_data = engagement_data.sort_index()

daily_engagement = engagement_data.index.value_counts()
```

```python
f, ax = plt.subplots(nrows=1, ncols=1, figsize=(24, 6))
plt.title('Daily engagement with products during the Covid-19 year in New-York')
plt.scatter(daily_engagement.index, daily_engagement)
plt.ylabel('Engagement')
plt.show()
```

* We can see here a volume of engagement with Educational Products during the Covid-19 first year, from 2020-30 till 2021-01
* We can see the drop in the engagement during the summer holidays 2020-07 till 2020-09
* We can see the hype at April 2020 (same as Covid-19 spread)
* We can see a hype relax after the Summer vocation

# Which products were the most popular in deifferent regions?

```python
engagement_with_products = pd.DataFrame(engagement_data.value_counts('Name'))
```

```python
daily_engagement_in_city = pd.DataFrame(engagement_data[engagement_data['locale'] == 'city']).value_counts('Name')
daily_engagement_in_rural = pd.DataFrame(engagement_data[engagement_data['locale'] == 'rural']).value_counts('Name')
daily_engagement_in_suburb = pd.DataFrame(engagement_data[engagement_data['locale'] == 'suburb']).value_counts('Name')
```

```python
# Normilize by total amount
daily_engagement_in_city = round(daily_engagement_in_city / daily_engagement_in_city.sum() * 10000)
daily_engagement_in_rural = round(daily_engagement_in_rural / daily_engagement_in_rural.sum() * 10000)
daily_engagement_in_suburb = round(daily_engagement_in_suburb / daily_engagement_in_suburb.sum() * 10000)
```

```python
for item in daily_engagement_in_city.iteritems():
    engagement_with_products.at[item[0], 'city_eng'] = item[1]

for item in daily_engagement_in_rural.iteritems():
    engagement_with_products.at[item[0], 'rural_eng'] = item[1]    
  
for item in daily_engagement_in_suburb.iteritems():
    engagement_with_products.at[item[0], 'suburb_eng'] = item[1]   
    
```

```python
engagement_with_top_30_products = engagement_with_products[:30]
```

```python
top_30_daily_engagement_in_city = engagement_with_top_30_products['city_eng']
top_30_daily_engagement_in_rural = engagement_with_top_30_products['rural_eng']
top_30_daily_engagement_in_suburb = engagement_with_top_30_products['suburb_eng']
```

```python
f, ax = plt.subplots(nrows=1, ncols=1, figsize=(14, 12))

plt.barh(top_30_daily_engagement_in_city.index, top_30_daily_engagement_in_city)
plt.barh(top_30_daily_engagement_in_rural.index, top_30_daily_engagement_in_rural)
plt.barh(top_30_daily_engagement_in_suburb.index, top_30_daily_engagement_in_suburb)
plt.legend(['city', 'rural', 'suburb'])
plt.show()
```

Seems there are some differences with Educational Products used in deifferent regiouns:

* Google Classroom is extreemly popular in City & Suburb, but not in Rural
* Google sites, TeachersPayTeachers, CNN student news  very popular Rural & Suburb, but not in City
* FlipGrid is widely used in City & Suburb, but not in Rural

# Focus on 3 products

```python
product = 'Google Classroom'
```

```python
engagement_with_product = pd.DataFrame(engagement_data[engagement_data['Name'] == product])

engagement_with_product_city = engagement_with_product[engagement_with_product['locale'] == 'city']
engagement_with_product_rural = engagement_with_product[engagement_with_product['locale'] == 'rural']
engagement_with_product_suburb = engagement_with_product[engagement_with_product['locale'] == 'suburb']

engagement_with_product_city = engagement_with_product_city.resample('W').agg({'engagement_index': 'max'})
engagement_with_product_rural = engagement_with_product_rural.resample('W').agg({'engagement_index': 'max'})
engagement_with_product_suburb = engagement_with_product_suburb.resample('W').agg({'engagement_index': 'max'})

f, ax = plt.subplots(nrows=1, ncols=1, figsize=(24, 6))

plt.plot(engagement_with_product_city.index, engagement_with_product_city)
plt.plot(engagement_with_product_rural.index, engagement_with_product_rural)
plt.plot(engagement_with_product_suburb.index, engagement_with_product_suburb)
plt.title('Engagement with '+ str(product))
plt.ylabel('engagement')
plt.legend(['city', 'rural', 'suburb'])
plt.show()
```
