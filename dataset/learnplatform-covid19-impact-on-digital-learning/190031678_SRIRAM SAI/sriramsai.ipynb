```python
import matplotlib.pyplot as plt
df_null = pd.DataFrame(df1.isnull().sum()/len(df1)).reset_index().sort_values(by=0)
plt.figure(figsize=(8,6))
plt.bar(df_null['index'],df_null[0],alpha=0.75)
plt.xticks(rotation=90)
plt.rc('ytick', labelsize=10)
plt.title('Percetage of missing values in products data')
```

```python
df2_null = pd.DataFrame(df2.isnull().sum()/len(df2)).reset_index().sort_values(by=0)
plt.figure(figsize=(8,6))
plt.bar(df2_null['index'],df2_null[0],alpha=0.75)
plt.xticks(rotation=90)
plt.rc('ytick', labelsize=10)
plt.title('Percetage of missing values in products data')
```
