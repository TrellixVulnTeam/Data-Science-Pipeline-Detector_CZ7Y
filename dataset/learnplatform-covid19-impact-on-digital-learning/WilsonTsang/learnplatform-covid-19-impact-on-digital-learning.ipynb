```python
#https://www.kaggle.com/yosefwalemneh/learnplatform-covid-19-impact-on-digital-learning

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
```

```python
import glob

eng_path = '../input/learnplatform-covid19-impact-on-digital-learning/engagement_data'
eng_files = glob.glob(eng_path + "/*.csv")

files = []

for file in eng_files:
    df = pd.read_csv(file, index_col = None, header = 0)
    district_id = file.split('/')[4].split('.')[0]
    df['district_id'] = district_id
    files.append(df)
    
engagement = pd.concat(files)
engagement = engagement.reset_index(drop = True)
engagement['time'] = pd.to_datetime(engagement['time'])
engagement['engagement_index'] = engagement['engagement_index'].fillna(0)

#engagement.to_csv('combined_df_1.csv')
#engagement.shape
```

```python
products = pd.read_csv('../input/learnplatform-covid19-impact-on-digital-learning/products_info.csv')
products.dropna(inplace = True)
districts = pd.read_csv('../input/learnplatform-covid19-impact-on-digital-learning/districts_info.csv')
districts.dropna(inplace = True)

COVID_cases = pd.read_csv('../input/usa-covid-case/usa_case.csv')
COVID_cases['date'] = pd.to_datetime(COVID_cases['date'])
COVID_cases['year'], COVID_cases['month'] = COVID_cases['date'].dt.year, COVID_cases['date'].dt.month

#products.to_csv('products.csv')
#districts.to_csv('districts.csv')
```

```python
used_app = engagement['lp_id'].unique()
all_app = products['LP ID'].unique()
used_app.shape
#pd.DataFrame(all_app).to_csv('all_app.csv')
#pd.DataFrame(used_app).to_csv('used_app.csv')
```

```python
#notuse_app = set(all_app) - set(used_app)
#print(notuse_app)
used_product = products[products['LP ID'].isin(used_app)]
engagement = engagement.rename(columns={"lp_id": "LP ID"})
app_engage = pd.merge(left=engagement,right=used_product,on='LP ID')
```

```python
#test = app_engage.head(100)
app_engage['app_category'] = app_engage['Primary Essential Function'].str.split(' - ',expand=True)[0]
app_engage['app_subcategory'] = app_engage['Primary Essential Function'].str.split(' - ',expand=True)[1]

app_engage['time'] = pd.to_datetime(app_engage['time'])
app_engage['year'], app_engage['month'] = app_engage['time'].dt.year, app_engage['time'].dt.month

app_engage.head(5)
```

```python
#pct_subcat = app_engage.groupby(['month','app_subcategory']).agg({'pct_access': ['mean', 'max']})
#engage_subcat = app_engage.groupby(['month','app_subcategory']).agg({'engagement_index': ['mean', 'max']})
```

```python
pct_subcat = pd.DataFrame(app_engage.groupby(['month','app_category']).agg({'pct_access': ['mean', 'max']})).reset_index()

engage_subcat = pd.DataFrame(app_engage.groupby(['month','app_category']).agg({'engagement_index': ['mean', 'max']})).reset_index()

```

```python
pct_subcat.columns = ['Month','category','Mean_pct','Max_pct']
engage_subcat.columns = ['Month','category','Mean_pct','Max_pct']
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.pyplot import figure
```

```python
def graph_format():
    plt.legend(loc="lower right")
    figure(figsize=(8, 8), dpi=100)
```
