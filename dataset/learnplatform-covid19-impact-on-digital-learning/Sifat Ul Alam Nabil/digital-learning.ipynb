```python
districts_data = pd.read_csv("../input/learnplatform-covid19-impact-on-digital-learning/districts_info.csv")
product_data = pd.read_csv("../input/learnplatform-covid19-impact-on-digital-learning/products_info.csv")
```

```python
districts_data = districts_data[districts_data.state.notna()].reset_index(drop=True)
districts_data.head()
```

```python
engagement_data_path = "../input/learnplatform-covid19-impact-on-digital-learning/engagement_data/"
temp = []
districts = districts_data.district_id.unique()
for district in districts: 
    temp_df = pd.read_csv(f"{engagement_data_path}/{district}.csv", index_col=None, header=0)
    temp_df["district_id"] = district
    temp.append(temp_df)
    
engagement_data = pd.concat(temp)
engagement_data = engagement_data.reset_index(drop=True)
engagement_data.head()
```

```python
engagement_data = pd.merge(engagement_data, product_data, left_on="lp_id", right_on="LP ID", how="left")
engagement_data.drop(["LP ID", "lp_id"], inplace=True, axis=1)

engagement_data["time"] = pd.to_datetime(engagement_data["time"])

engagement_data["month"] = engagement_data["time"].dt.month_name()

engagement_data.head()
```
