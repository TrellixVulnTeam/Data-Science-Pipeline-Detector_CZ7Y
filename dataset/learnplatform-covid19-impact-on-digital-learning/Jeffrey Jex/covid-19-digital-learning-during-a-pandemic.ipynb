```python
state_count = district_df.groupby(["state", "locale"]).count()
state_count = state_count[['district_id']]
state_count.columns = ["district_count"]
state_count.reset_index(inplace=True)
state_count.sort_values(by="district_count", ascending = False, inplace=True)
px.bar(state_count, x="state", y="district_count", color="locale", title="Number of participating districts, by state")
```
