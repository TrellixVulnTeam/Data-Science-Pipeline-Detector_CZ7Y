## In this naive sample submission, we submit the last grid as the initial grid, naively assuming therefore that there was no change. The solely purpose was to get an idea of the worst score possibly achievable and an idea of performances range.

```python
import numpy as np
import pandas as pd
```

```python
sub = pd.read_csv("../input/conways-reverse-game-of-life-2020/test.csv")
```

```python
INIT_COLS = []
STOP_COLS = []
REN_DICT = {}
for i in range(625):
    INIT_COLS.append(f"start_{i}")
    STOP_COLS.append(f"stop_{i}")
    REN_DICT[f"stop_{i}"] = f"start_{i}"
#=====
sub = sub.rename(columns=REN_DICT) # swapping stopping columns as starting ones
sub.drop("delta", axis=1, inplace=True)
```
