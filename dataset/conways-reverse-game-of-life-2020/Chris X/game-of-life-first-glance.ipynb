# First look on the data

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import matplotlib.pyplot as plt

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

# Training Data

```python
# read training data
df_train = pd.read_csv('../input/conways-reverse-game-of-life-2020/train.csv')
# show dimensions
df_train.shape
```

```python
df_train.head()
```

```python
df_train.delta.value_counts().plot(kind='bar')
plt.title('Distribution of delta')
plt.grid()
plt.show()
```

### Check transitions from start to end

```python
pd.crosstab(df_train.start_0, df_train.stop_0)
```

```python
pd.crosstab(df_train.start_1, df_train.stop_1)
```

```python
pd.crosstab(df_train.start_624, df_train.stop_624)
```

### Select an example and display configurations

```python
# select an example
sel_row = 2
example = df_train.iloc[sel_row,:]
example
```

```python
# extract start values
start = example[2:2+625]
start
```

```python
# convert to matrix
start = np.asarray(start)
start_mat = start.reshape(25,25)
```

```python
# extract stop values
stop = example[2+625:2+625+625+1]
stop
```

```python
# convert to matrix
stop = np.asarray(stop)
stop_mat = stop.reshape(25,25)
```

```python
# show start configuration
plt.imshow(start_mat)
plt.show()
```

```python
# show stop configuration
plt.imshow(stop_mat)
plt.show()
```

# Test data

```python
# read test data
df_test = pd.read_csv('../input/conways-reverse-game-of-life-2020/test.csv')
# show dimensions
df_test.shape
```

```python
df_test.head()
```

```python
# count active cells per row
df_test_for_sum = df_test.copy()
df_test_for_sum = df_test_for_sum.drop(['id','delta'], axis=1)
active_count = df_test_for_sum.sum(axis=1)
active_count
```

```python
# show distribution of number of active cells
plt.hist(active_count.values,50)
plt.grid()
plt.title('Number of active cells')
plt.show()
```

### Select an example and display

```python
# select an example
sel_row = 3
example = df_test.iloc[sel_row,:]
example
```

```python
# extract stop values
stop = example[2:2+625]
# convert to matrix
stop = np.asarray(stop)
stop_mat = stop.reshape(25,25)
```
