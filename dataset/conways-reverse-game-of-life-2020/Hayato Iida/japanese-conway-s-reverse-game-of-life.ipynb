# コンペについて
- コンウェイのライフゲームの事前状態を機械学習で予想する
- 25×25のサイズ
- 上端と下端がつながっている(トーラス)
- 初期状態と最終状態、最終状態に至るまでのステップ数が与えられる
- 50000パターンのデータが提供されているが当然いくらでも生成可能
- 完全にランダムに生成されているわけではなく以下のステップでボードの選定が行われている
  
  
1. 1%~99%の間でランダムな密度でボードを作成
2. 5 step 処理を進めこれを初期状態とする
3. 1~5の間でランダムにdeltaを決定
4. 初期状態からdelta step分進める
5. 終了状態でボードが空の場合破棄する

# コンウェイのライフゲーム
- 以下の4つのルールで構成される
- 誕生
  - 死んでいるセルに隣接する生きたセルがちょうど3つあれば、次の世代が誕生する。
- 生存
  - 生きているセルに隣接する生きたセルが2つか3つならば、次の世代でも生存する。
- 過疎
  - 生きているセルに隣接する生きたセルが1つ以下ならば、過疎により死滅する。
- 過密
  - 生きているセルに隣接する生きたセルが4つ以上ならば、過密により死滅する。
  
 (https://ja.wikipedia.org/wiki/%E3%83%A9%E3%82%A4%E3%83%95%E3%82%B2%E3%83%BC%E3%83%A0)
