### There are four simple rules to determine the next state of the game board, given the current state:

1. Overpopulation: if a living cell is surrounded by more than three living cells, it dies.
1. Stasis: if a living cell is surrounded by two or three living cells, it survives.
1. Underpopulation: if a living cell is surrounded by fewer than two living cells, it dies.
1. Reproduction: if a dead cell is surrounded by exactly three cells, it becomes a live cell.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
```

```python
train_df = pd.read_csv('../input/conways-reverse-game-of-life-2020/train.csv')
test_df = pd.read_csv("../input/conways-reverse-game-of-life-2020/test.csv")
print(train_df.shape)
print(test_df.shape)
```

```python
train_df.head()
```

```python
colors = ['white','black'] 
bounds = [0,1] 
cmap = mpl.colors.ListedColormap(colors)
norm = mpl.colors.BoundaryNorm(bounds,cmap.N)
for i in range(3):
    # convert to matrix
    start = np.asarray(train_df.iloc[i][2:2+625])
    start_mat = start.reshape(25,25)

    # extract stop values
    stop = np.asarray(train_df.iloc[i][2+625:])
    stop_mat = stop.reshape(25,25)
    
    plt.rcParams['figure.figsize'] = (15.0, 7.0) #  Set the figsize
    plt.figure()
    plt.subplot(1,2,1)
    plt.xlabel('start_map')
    plt.imshow(start_mat,cmap = cmap)
    plt.colorbar()

    plt.subplot(1,2,2)
    plt.xlabel('stop_map')
    plt.imshow(stop_mat,cmap = cmap)
    plt.colorbar()
    plt.show()
```

### Let's try to add/minus them to find something between the begin and stop map.

```python
colors = ['black','white','yellow','blue'] 
bounds = [-1,0,1,2] 
cmap = mpl.colors.ListedColormap(colors)
norm = mpl.colors.BoundaryNorm(bounds,cmap.N)
for i in range(3):
    # convert to matrix
    start = np.asarray(train_df.iloc[i][2:2+625])
    start_mat = start.reshape(25,25)
    # extract stop values
    stop = np.asarray(train_df.iloc[i][2+625:])
    stop_mat = stop.reshape(25,25)
    
    plt.rcParams['figure.figsize'] = (15.0, 7.0) #  Set the figsize
    plt.figure()
    plt.subplot(1,2,1)
    plt.xlabel('add')
    plt.imshow(start_mat+stop_mat,cmap = cmap,norm = norm)
    plt.colorbar()

    plt.subplot(1,2,2)
    plt.xlabel('minus')
    plt.imshow(start_mat-stop_mat,cmap = cmap,norm = norm)
    plt.colorbar()
    plt.show()
    
```
