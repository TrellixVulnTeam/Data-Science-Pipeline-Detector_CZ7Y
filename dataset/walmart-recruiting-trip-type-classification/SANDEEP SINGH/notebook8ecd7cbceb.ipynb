```python
def flot_to_str(obj):
    """
    Convert Upc code from float to string.
    Use this function by applying lambda
    Parameters: "Upc" column of DataFrame
    Return:string converted Upc removing dot
    """
    while obj != 'np.nan':
        obj = str(obj).split('.')[0]
        if len(obj) == 10:
            obj = obj + '0'
        elif len(obj) == 4:
            obj = obj + '0000000' 
        return obj
```

```python
def prodct(upcData):
    """
    Return company code from given Upc code.
    Parameters:'Upc' column of DataFrame
    Return: company code
    """
    try:
        code = upcData[6 :]
        return code
    except:
        return -9999
```
