## Creación del DataFrame

```python
df = pd.read_csv(directorio + entrenamiento)
df
```

***En principio, el dataframe contiene 647054 filas y 7 columnas***. 

Cada una de las columnas representa lo siguiente, de acuerdo a la descripción:

```python
df.columns
```

- **TripType**: representa el tipo de viaje de compras que realizó el cliente. Ésta variable es la que se está prediciendo. TripType_999 es una categoría "otra".
- **VisitNumber**: identificación correspondiente a un solo viaje de un solo cliente.
- **Weekday**: el día de la semana del viaje del cliente.
- **UPC**: número UPC del producto comprado por el cliente.
- **ScanCount**: número del artículo que se compró. Un valor negativo indica una devolución de producto.
- **DepartmentDescription**: departamento donde pertenece el artículo que se compro.
- **FinelineNumber**: categoría de los productos de Walmart.
