```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
import os, glob

drivers = pd.read_csv('../input/driver_imgs_list.csv')
train_files = [f for f in glob.glob("../input/train/*/*.jpg")]
test_files = ["../input/test/" + f for f in os.listdir("../input/test/")]
print(train_files[:10])
print(test_files[:10])
```

```python
plt.rcParams['figure.figsize'] = (12.0, 12.0)
plt.subplots_adjust(wspace=0, hspace=0)

c_files = []
c_files.append(['/usr/local/share/OpenCV/haarcascades/haarcascade_eye.xml','eye'])
c_files.append(['/usr/local/share/OpenCV/haarcascades/haarcascade_eye_tree_eyeglasses.xml','glasses'])
c_files.append(['/usr/local/share/OpenCV/haarcascades/haarcascade_frontalcatface.xml','frontal'])
c_files.append(['/usr/local/share/OpenCV/haarcascades/haarcascade_frontalcatface_extended.xml','cat ext'])
c_files.append(['/usr/local/share/OpenCV/haarcascades/haarcascade_frontalface_alt.xml','alt'])
c_files.append(['/usr/local/share/OpenCV/haarcascades/haarcascade_frontalface_alt2.xml','alt2'])
c_files.append(['/usr/local/share/OpenCV/haarcascades/haarcascade_frontalface_alt_tree.xml','alt tree'])
c_files.append(['/usr/local/share/OpenCV/haarcascades/haarcascade_frontalface_default.xml','default'])
c_files.append(['/usr/local/share/OpenCV/haarcascades/haarcascade_fullbody.xml','body'])
c_files.append(['/usr/local/share/OpenCV/haarcascades/haarcascade_lefteye_2splits.xml','splits'])
c_files.append(['/usr/local/share/OpenCV/haarcascades/haarcascade_licence_plate_rus_16stages.xml','license'])
c_files.append(['/usr/local/share/OpenCV/haarcascades/haarcascade_lowerbody.xml','lowerbody'])
c_files.append(['/usr/local/share/OpenCV/haarcascades/haarcascade_profileface.xml','profile'])
c_files.append(['/usr/local/share/OpenCV/haarcascades/haarcascade_righteye_2splits.xml','right eye'])
c_files.append(['/usr/local/share/OpenCV/haarcascades/haarcascade_russian_plate_number.xml','russian'])
c_files.append(['/usr/local/share/OpenCV/haarcascades/haarcascade_smile.xml','smile'])
c_files.append(['/usr/local/share/OpenCV/haarcascades/haarcascade_upperbody.xml','upper'])
c_files.append(['/usr/local/share/OpenCV/lbpcascades/lbpcascade_frontalcatface.xml','cat frontal'])
c_files.append(['/usr/local/share/OpenCV/lbpcascades/lbpcascade_frontalface.xml','lbp frontal'])
c_files.append(['/usr/local/share/OpenCV/lbpcascades/lbpcascade_profileface.xml','lbp profile'])
c_files.append(['/usr/local/share/OpenCV/lbpcascades/lbpcascade_silverware.xml','silver'])

import random
fi = random.choice(train_files)
print(fi)
im = cv2.imread(fi)
plt.imshow(cv2.cvtColor(im, cv2.COLOR_BGR2RGB)); plt.axis('off')
```

```python
i_ = 0
plt.rcParams['figure.figsize'] = (11.0, 21.0)
plt.subplots_adjust(wspace=0, hspace=0)
for c in c_files:
    im2 = im.copy()
    gr_im = cv2.cvtColor(im2, cv2.COLOR_BGR2GRAY)
    fc = cv2.CascadeClassifier(c[0])
    fr = fc.detectMultiScale(gr_im, scaleFactor=1.1, minNeighbors=2, minSize=(20, 20), flags = cv2.CASCADE_SCALE_IMAGE)
    if len(fr)>0:
        for (x, y, w, h) in fr:
            cv2.rectangle(im2, (x, y), (x+w, y+h), (0, 0, 255), 2)
        cv2.cvtColor(im2, cv2.COLOR_BGR2RGB)
    plt.subplot(7, 3, i_+1).set_title(c[1])
    plt.imshow(cv2.cvtColor(im2, cv2.COLOR_BGR2RGB)); plt.axis('off')
    i_ += 1
```
