```python
x_train = []
y_train = []

def train_data(lab,path):
    folders = glob.glob(path)
    imagenames_list = []
    for folder in folders:
        for f in glob.glob(folder+'/*.jpg'):
            imagenames_list.append(f)


    for image in imagenames_list:
        im = cv2.imread(image, cv2.IMREAD_GRAYSCALE)
        scale_percent = 60 # percent of original size
        width = int(im.shape[1] * scale_percent / 100)
        height = int(im.shape[0] * scale_percent / 100)
        dim = (width, height)
        im = cv2.resize(im, dim)
        x_train.append(im)
        y_train.append(lab)
```

```python
x_train.clear()
y_train.clear()
for sub in os.listdir('../input/state-farm-distracted-driver-detection/imgs/train'): 
    train_data(sub,os.path.join('../input/state-farm-distracted-driver-detection/imgs/train/',sub))
```
