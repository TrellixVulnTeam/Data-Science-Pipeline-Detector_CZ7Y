```python
train_2020 = pd.read_csv('../input/hpa-single-cell-image-classification/train.csv')
train_2018 = pd.read_csv('../input/human-protein-atlas-image-classification/train.csv')
```

```python
train_2018.rename(columns={'Id':'ID'}, inplace=True)
```

```python
train_merged = train_2020.merge(train_2018, on='ID', how='inner')
train_merged
```

```python
label_2020 = pd.DataFrame()
tmp = train_merged['Label'].map(lambda x: list(map(int, x.split('|'))))
for i in range(19):
    label_2020[i] = tmp.map(lambda x: i in x) * 1

label_2020
```

```python
label_2018 = pd.DataFrame()
tmp = train_merged['Target'].map(lambda x: list(map(int, x.split())))
for i in range(19):
    label_2018[i] = tmp.map(lambda x: i in x) * 1

label_2018
```
