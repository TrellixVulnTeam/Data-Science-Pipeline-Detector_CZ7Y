# Tutorial: Installing dependencies for offline submissions

This notebook shows how to install dependencies to notebooks for offline submission (this may be obvious to Kaggle veterans).

```python
# This example notebook contains pip packages, binaries, or pre-trained models to be used in offline notebooks.
# After you run this, do File -> Save version -> Advanced Settings -> Save output for this version -> Quick save -> Save.
# On the "inference" notebook, do File -> Add or upload data -> Notebook output files -> Your work, find this notebook, and "Add" it.
# It may take a few minutes for it to show up on the list.
# These files will then be available on /kaggle/input/ on the "inference" notebook, and show up under "data" on the right pane.

!mkdir -p /kaggle/working/pip
!pip download --exists-action=w kornia==0.6.4 -d /kaggle/working/pip/kornia
!pip download --exists-action=w kornia_moons -d /kaggle/working/pip/kornia_moons
!pip download --exists-action=w git+https://github.com/cvlab-epfl/disk@kaggle -d /kaggle/working/pip/disk

!mkdir -p wheels
!pip wheel -w wheels git+https://github.com/jatentaki/unets.git@no_norm_at_start
!pip wheel -w wheels git+https://github.com/jatentaki/torch-localize.git@flatten
!pip wheel -w wheels git+https://github.com/jatentaki/torch-dimcheck
!pip wheel -w wheels git+https://github.com/etrulls/disk@kaggle

!mkdir -p /kaggle/working/pretrained
!wget http://cmp.felk.cvut.cz/~mishkdmy/hardnet8v2.pt -O /kaggle/working/pretrained/hardnet8v2.pt
!wget https://github.com/axelBarroso/Key.Net-Pytorch/raw/main/model/weights/keynet_pytorch.pth -O /kaggle/working/pretrained/keynet_pytorch.pth
!wget https://github.com/ducha-aiki/affnet/raw/master/pretrained/AffNet.pth -O /kaggle/working/pretrained/AffNet.pth
!wget https://github.com/ducha-aiki/affnet/raw/master/pretrained/OriNet.pth -O /kaggle/working/pretrained/OriNet.pth


!wget https://github.com/etrulls/disk/raw/master/depth-save.pth -O /kaggle/working/pretrained/disk-depth.pth
!wget https://github.com/etrulls/disk/raw/master/epipolar-save.pth -O /kaggle/working/pretrained/disk-epipolar.pth
!wget http://cmp.felk.cvut.cz/~mishkdmy/models/loftr_outdoor.ckpt -O /kaggle/working/pretrained/loftr_outdoor.ckpt
!wget https://github.com/ducha-aiki/Key.Net-Pytorch/raw/main/model/HyNet/weights/HyNet_LIB.pth -O /kaggle/working/pretrained/HyNet_LIB.pth
!wget https://github.com/DagnyT/hardnet/raw/master/pretrained/pretrained_all_datasets/HardNet++.pth /kaggle/working/pretrained/HardNet++.pth
!wget https://github.com/DagnyT/hardnet/raw/master/pretrained/train_liberty_with_aug/checkpoint_liberty_with_aug.pth /kaggle/working/pretrained/HardNet.pth



print('Done!')
```

```python
# Now you can flick the "offline" switch on the right pane and test the installation.

!pip install kornia --no-index --find-links=file:///kaggle/working/pip/kornia/ --upgrade
!pip install kornia_moons --no-index --find-links=file:///kaggle/working/pip/kornia_moons/ --no-deps  --upgrade
!pip install -f /kaggle/working/wheels --no-index torch_dimcheck
!pip install -f /kaggle/working/wheels --no-index torch_localize
!pip install -f /kaggle/working/wheels --no-index unets
!pip install -f /kaggle/working/wheels --no-index disk
print('Done!')
```
