```python
import numpy as np
import pandas as pd
```

```python
calibration = pd.read_csv('../input/image-matching-challenge-2022/train/brandenburg_gate/calibration.csv')
pair = pd.read_csv('../input/image-matching-challenge-2022/train/brandenburg_gate/pair_covisibility.csv')
pair.head()
```

```python
_,k1,r1,t1 = calibration[calibration.image_id=='90920828_5082887495'].values[0]
k1 = [eval(x) for x in k1.split()]
r1 = [eval(x) for x in r1.split()]
t1 = [eval(x) for x in t1.split()]
k1, r1, t1
```

```python
_,k2,r2,t2 = calibration[calibration.image_id=='20133057_3035445116'].values[0]
k2 = [eval(x) for x in k2.split()]
r2 = [eval(x) for x in r2.split()]
t2 = [eval(x) for x in t2.split()]
k2, r2, t2
```

```python
def get_intrinsics(k):
    return np.array(k).reshape(3, 3)

def get_extrinsics(r, t):
    R = np.array(r).reshape(3, 3)
    T = np.array(t)
    extrinsic = np.eye(4)
    extrinsic[:3, :3] = R
    extrinsic[:3, 3] = T
    return extrinsic

def skew(x):
    return np.array([[0, -x[2], x[1]],
                     [x[2], 0, -x[0]],
                     [-x[1], x[0], 0]])
```

```python
intrinsic1 = get_intrinsics(k1)
intrinsic2 = get_intrinsics(k2)

extrinsic1 = get_extrinsics(r1,t1)
extrinsic2 = get_extrinsics(r2,t2)

relative = extrinsic2.dot(np.linalg.inv(extrinsic1))
relative
```

```python
R = relative[:3, :3]
T = relative[:3, 3]
tx = skew(T)
E = np.dot(tx, R)
F = np.linalg.inv(intrinsic2).T.dot(E).dot(np.linalg.inv(intrinsic1))
F
```

```python
F_gt = pair[pair['pair']=='90920828_5082887495-20133057_3035445116'].fundamental_matrix.values[0]
F_gt = [eval(x) for x in F_gt.split()]
F_gt = np.array(F_gt).reshape(3,3)
F_gt
```
