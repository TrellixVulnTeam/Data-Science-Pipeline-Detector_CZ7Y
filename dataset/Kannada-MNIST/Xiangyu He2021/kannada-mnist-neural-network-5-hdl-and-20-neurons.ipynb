```python
import pandas as pd
import time as time

train=pd.read_csv('../input/Kannada-MNIST/train.csv')
train.head()

test=pd.read_csv('../input/Kannada-MNIST/test.csv')


idtest = test["id"]
xtest=test.drop(['id'],axis=1)
xtest.head()
```

```python
label=train['label']
xtrain=train.drop(['label'],axis=1)
from sklearn.model_selection import train_test_split
x_train,x_val,y_train,y_val=train_test_split(xtrain,label,test_size=0.2,random_state=1)#converting the dataset into proper LGB format 
```

# EDA using PCA

```python
trainval = pd.concat([x_train,x_val],axis=0)
traintest = pd.concat([xtrain,xtest],axis=0)
from sklearn.decomposition import PCA
start=time.time()
pca = PCA(n_components= 240)
pca.fit(trainval)
pca.fit(traintest)
datanew1 = pca.fit_transform(trainval)
datanew2 = pca.fit_transform(traintest)
end=time.time()
print(end-start)
print(sum(pca.explained_variance_ratio_))
datanew1.shape
datanew2.shape
```

```python
import numpy as np
data_train = datanew1[:len(x_train),:]
data_val = datanew1[len(x_train):,:]
data_test = datanew2[len(xtrain):,:]
print(data_train.shape,data_val.shape,data_test.shape)
```

# Neural network model 1: 2 Hidden layers 10 neurons

```python
from sklearn.neural_network import MLPClassifier
nn210 = MLPClassifier(solver='sgd', alpha=1e-5,
                    hidden_layer_sizes=(2, 10), random_state=1)
start=time.time()
nn210.fit(data_train, y_train)
end=time.time()

from sklearn.metrics import accuracy_score

ACCtrain=accuracy_score(y_train,nn210.predict(data_train))*100
ACCtest=accuracy_score(y_val,nn210.predict(data_val))*100
print('the fit time is',end-start,
      'the train accuracy is :',ACCtrain,
      'the test accuracy is :',ACCtest)

submit = pd.DataFrame({"id":idtest, "label": nn210.predict(data_test)})
submit.to_csv("submission.csv", index = False)
```
