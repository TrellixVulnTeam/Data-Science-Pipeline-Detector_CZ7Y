```python
train=pd.read_csv('../input/Kannada-MNIST/train.csv')
test=pd.read_csv('../input/Kannada-MNIST/test.csv')
submission = pd.read_csv('../input/Kannada-MNIST/sample_submission.csv')
```

```python
X, y = (train.drop(['label'], axis=1), train.label)
```

```python
X = X / 255
X.max().max()
```

```python
test = test / 255
```

```python
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
svc = svm.SVC(kernel = "rbf", C=4, cache_size=1000)
fitted = svc.fit(X, y)
```

```python
trainPred = fitted.predict(X)

print("Final training score for XGB: ", accuracy_score(y, trainPred))

test_x = test.values[:,1:]
xgbTestPred = fitted.predict(test_x)
```

```python
submission['label'] = xgbTestPred
submission.to_csv('submission.csv', index=False)
submission.head()
```
