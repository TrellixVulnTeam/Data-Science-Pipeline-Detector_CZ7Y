```python
# essential tools
import pandas as pd
import numpy as np
import pickle
from sklearn.ensemble import RandomForestClassifier
from datetime import datetime

data = pd.read_csv('/kaggle/input/Kannada-MNIST/train.csv')
X = data[data.columns[1:]]
y = data[data.columns[0]]
randomforest1 = RandomForestClassifier(n_estimators=10,
                                      max_features='sqrt',
                                      bootstrap=True,
                                      n_jobs=1,
                                      random_state=0)
start=datetime.now() # 2021-04-30 16:42:43.157194
randomforest1.fit(X,y)
end=datetime.now() # 2021-04-30 16:42:48.094273
print(end-start) # 0:00:04.937079

test_data = pd.read_csv('/kaggle/input/Kannada-MNIST/test.csv')
test_ids = test_data[test_data.columns[0]]
X_test = test_data[test_data.columns[1:]]

test_predictions = randomforest1.predict(X_test)
test_predictions = pd.DataFrame({'id': test_ids,
                                 'label': test_predictions})

test_predictions.to_csv('submission.csv',
                        header=['id', 'label'],
                        index=False)
```
