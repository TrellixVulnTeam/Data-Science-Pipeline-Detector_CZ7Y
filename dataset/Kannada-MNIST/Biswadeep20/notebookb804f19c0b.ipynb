```python


import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)


test_dataframe=pd.read_csv('../input/Kannada-MNIST/test.csv')
train_dataframe=pd.read_csv('../input/Kannada-MNIST/train.csv')
```

```python
train_dataframe.head()
```

```python
train_labels=train_dataframe.pop('label')
train_labels.head()
```

```python
train_input=train_dataframe
train_input.head()
```

```python
train_input.isnull().any().describe()
```

```python
test_dataframe.pop("id")

test_dataframe
```

```python
train_input=train_input/255.0
```

```python

test_input = test_dataframe/255.0

```

```python
train_input=train_input.values.reshape(-1,28,28,1)
test_dataframe=test_input.values.reshape(-1,28,28,1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_labels.shape
```

```python
train_set,val_set,train_label,val_label=train_test_split(train_input,train_labels,test_size=0.1)
```

```python
import matplotlib.pyplot as plt
plt.imshow(test_dataframe[53][:,:,0])
```

```python
import tensorflow as tf

datagen=tf.keras.preprocessing.image.ImageDataGenerator(
    rotation_range=5,
    zoom_range = 0.2,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    fill_mode="nearest"
    
)
```

```python
datagen.fit(train_set)

train_datagen=datagen.flow(train_set,train_label,batch_size=400)
```

```python
model=tf.keras.Sequential([
    
  tf.keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(28, 28, 1)),
  tf.keras.layers.MaxPooling2D(2, 2),
  tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
  tf.keras.layers.MaxPooling2D(2,2),
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dense(512, activation='relu'),
  tf.keras.layers.Dropout(0.5),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dense(10, activation='softmax')
    
])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
history=model.fit(train_set,train_label,epochs=20,validation_data=(val_set,val_label),steps_per_epoch=70)
```

```python
accu=history.history['accuracy']
loss=history.history['loss']
val_accu=history.history['val_accuracy']
val_loss=history.history['val_loss']
epoch_num=range(len(accu))
plt.plot(epoch_num,accu,'r',label='accuracy')
plt.plot(epoch_num,loss,'b',label='loss')
plt.legend()
plt.figure()

plt.plot(epoch_num,val_accu,'r',label='val_accuracy')
plt.plot(epoch_num,val_loss,'b',label='val_loss')
plt.legend()
plt.figure()
plt.show()
```

```python
y_pre=model.predict(test_dataframe)     ##making prediction
y_pre=np.argmax(y_pre,axis=1) ##changing the prediction intro labels
```

```python
sample_sub=pd.read_csv('../input/Kannada-MNIST/sample_submission.csv')

sample_sub['label']=y_pre
sample_sub.to_csv('submission.csv',index=False)

sample_sub.head()
```
