```python
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
def get_data(data_set, label_column='label'):
    # data_set:   pd DataFrame
    data_X = data_set.drop(columns=label_column)
    data_y = data_set[label_column]
    return data_X, data_y

def count_labels(data_set, label_column='label'):
    # count the number of entries of each label
    labels = np.unique(data_set[label_column])
    for i in range(len(labels)):
        count_labels = data_set[data_set['label'] == i]
        print(f'laebl: {i}, count: {len(count_labels)}')    
    
```

```python
# read data
train_set = pd.read_csv('./train.csv')
test_set = pd.read_csv('./test.csv')
dig_set = pd.read_csv('./Dig-MNIST.csv')
```

```python
X_train, y_train = get_data(train_set, label_column='label')
X_dig, y_dig = get_data(dig_set, label_column='label')
```

```python
# check homogeneity of training set
count_labels(train_set)
```

```python
# check homogeneity of validation set
count_labels(dig_set)
```

```python
# plt.imshow(train_X.loc[1].to_numpy().reshape(28, 28))
# plt.show()
```

```python
for k in [5, 50, 100, 250, 500, 800, 1000]:
    knn_clf = KNeighborsClassifier(n_neighbors=k)
    knn_clf.fit(X=X_train, y=y_train)
    y_pred = knn_clf.predict(X_dig)
    accur = accuracy_score(y_dig, y_pred)
    print(f'k: {k}, accuracy: {accur}')
```
