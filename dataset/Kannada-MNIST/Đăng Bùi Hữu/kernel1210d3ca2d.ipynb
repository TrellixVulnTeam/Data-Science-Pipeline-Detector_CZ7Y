```python
train = pd.read_csv('/kaggle/input/digit-recognizer/train.csv')
test = pd.read_csv('/kaggle/input/digit-recognizer/test.csv')
```

```python
Y_train = train['label']
X_train = train.drop(labels='label',axis=1)

#Y_test = test['label']
#X_test = test.drop(labels='label',axis=1)
```

```python
# Normalize
X_train /= 255
test /= 255 # to do what?
#X_test /= 255
```

Преобразование X
Потому что рисунка сохраняется как последовательность
Сейчас нужно преобразовать в тип рисунки и потом возможно конвертировать

784px vectors to 28x28x1 3D matrices

```python
# Reshape
X_train = X_train.values.reshape(-1,28,28,1)
test = test.values.reshape(-1,28,28,1)

print(X_train.shape)
```

Преобразование Y поскольку  текущее значение Y 1-10, надо преобразовать на one_hot


```python
# to one_hot 
Y_train = to_categorical(Y_train, num_classes = 10)
```

```python
print ("number of training examples = " + str(X_train.shape[0]))
#print ("number of test examples = " + str(X_test.shape[0]))
print ("X_train shape: " + str(X_train.shape))
print ("Y_train shape: " + str(Y_train.shape))
#print ("X_test shape: " + str(X_test.shape))
#print ("Y_test shape: " + str(Y_test.shape))
```

```python
random_seed = 2
from sklearn.model_selection import train_test_split
X_train, X_val, Y_train, Y_val = train_test_split(X_train, Y_train, test_size = 0.1, random_state=random_seed)
```

                                  =====MODEL=====

```python
# the model

def MaiHaiModel(input_shape):
    X_input = Input(input_shape)
    X = ZeroPadding2D((3,3))(X_input) 
    X = Conv2D(32,(7,7),strides=(1,1),name='conv0')(X)
    X = BatchNormalization(axis=3, name='bn0')(X)
    X = Activation('relu')(X)
    
    X = MaxPooling2D((2,2), name='max_pool')(X)
    
    X = Flatten()(X)
    X = Dense(10, activation='softmax', name='fc')(X)
    
    model = Model(inputs= X_input, outputs = X, name='HapplyModel')
    
    return model
```

```python
maihaiModel = MaiHaiModel(X_train.shape[1:]) 
```

//optimizer = adam
//optimizer = RMSprop(lr=0.001, rho=0.9, epsilon=1e-08, decay=0.0)//




```python
maihaiModel.compile(optimizer='adam',loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
maihaiModel.fit(X_train,Y_train, epochs=1, batch_size=86)
```

```python
results = maihaiModel.predict(test)

results = np.argmax(results,axis = 1)

results = pd.Series(results,name="Label")
```
