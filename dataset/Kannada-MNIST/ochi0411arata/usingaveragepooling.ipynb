```python
import numpy as np
import pandas as pd
import os

train = pd.read_csv('../input/Kannada-MNIST/train.csv')
test = pd.read_csv('../input/Kannada-MNIST/test.csv')
print(f'trainデータ数：{len(train)}\n')
print(f'testデータ数：{len(test)}')
```

```python
train_y = train["label"]
train_x = train.drop(["label"],axis=1)
train_y,val_y,train_x,val_x = train_y[:50000],train_y[50000:],train_x[:50000],train_x[50000:]

test_y = test["id"]
test_x = test.drop(["id"],axis=1)
```

```python
train_x = np.array(train_x).reshape(-1,28,28,1)
val_x = np.array(val_x).reshape(-1,28,28,1)
test_x = np.array(test_x).reshape(-1,28,28,1)

train_x = train_x.astype('float32')
val_x = val_x.astype('float32')
test_x = test_x.astype('float32')

train_x /= 255
val_x /= 255
test_x /= 255
```

```python
from keras.utils.np_utils import to_categorical
train_y = to_categorical(train_y,num_classes=10)
val_y = to_categorical(val_y,num_classes=10)
train_y[0]
```

```python
from keras.models import Sequential,load_model
from keras.layers import Dense,Dropout,Flatten,Conv2D,MaxPooling2D

model = Sequential()
```

```python
model.add(Conv2D(32,kernel_size=(5,5),activation='relu',input_shape=(28,28,1)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(64,kernel_size=(5,5),activation='relu',input_shape=(28,28,1)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dense(128,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='sgd',metrics=['accuracy'])
```

```python
history = model.fit(train_x,train_y,batch_size=256,epochs=20,validation_data=(val_x,val_y))
```

```python
predict_value = model.predict(test_x)
predict_value2=[]
for item in predict_value:
    value2=np.argmax(item)
    predict_value2+=[value2]
```

```python
submit=pd.read_csv('../input/Kannada-MNIST/sample_submission.csv')
submit['label']=predict_value2
submit.to_csv("submission.csv",index=False)
```
