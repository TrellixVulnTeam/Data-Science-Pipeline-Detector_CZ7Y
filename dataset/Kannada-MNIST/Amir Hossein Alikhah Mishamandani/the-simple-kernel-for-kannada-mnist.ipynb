**Introduction**

This is a simple version of the code I've created for [Kannada MNIST](https://www.kaggle.com/c/Kannada-MNIST) it is a preactice for me to create kernels on Kaggle and both explain a simple approach.

**Importing required libraries**

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.metrics import  accuracy_score, classification_report, confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
import csv
import os
import warnings
warnings.filterwarnings('ignore')
import gc
gc.enable()
```

**Loading data**

```python
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

**Loading training data 60000*785 -> y_tr = [0:60000,0], X_tr = [0:60000,1:]**

```python
data = pd.read_csv('/kaggle/input/Kannada-MNIST/train.csv')
tr_data = np.array(data).astype('int')
X_tr = tr_data[0:60000,1:]
y_tr = tr_data[0:60000,0]
del data, tr_data
gc.collect()
```

**Loading train test data 10240*785 -> y_tst = [0:10240,0], X_tst = [0:10240,1:]**

```python
data = pd.read_csv('/kaggle/input/Kannada-MNIST/Dig-MNIST.csv')
tr_tst_data = np.array(data).astype('int')
X_tst = tr_tst_data[0:10240,1:]
y_tst = tr_tst_data[0:10240,0]
del data, tr_tst_data
gc.collect()
```

**Loading final test data 5000*785 -> X = [0:5000,1:], id = [0:5000,0]**

```python
data = pd.read_csv('/kaggle/input/Kannada-MNIST/test.csv')
fl_tst_data = np.array(data).astype('int')
X = fl_tst_data[0:5000,1:]
del data, fl_tst_data
gc.collect()
print("Data is ready !!")
```

**Selecting the Classifier**

```python
clf = SVC(kernel='rbf')
```

**Classification**

```python
clf.fit(X_tr, y_tr)
```

**Testing**

```python
predictions = clf.predict(X_tst)
```

**Accuracy**

```python
print('The accuracy: ', accuracy_score(y_tst, predictions))
print(confusion_matrix(y_tst,predictions))
print(classification_report(y_tst,predictions))
del predictions
gc.collect()
```

Final test

```python
predictions = clf.predict(X)
sub = pd.read_csv('/kaggle/input/Kannada-MNIST/sample_submission.csv')
sub['label'] = predictions
sub.to_csv('submission.csv', index=False)
```
