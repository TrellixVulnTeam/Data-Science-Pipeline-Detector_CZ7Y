```python
data = pd.read_csv("../input/Kannada-MNIST/train.csv")
data.head()
```

```python
data.shape
```

```python
data.columns
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import utils
```

```python
X = pd.DataFrame([data["pixel" + str(i)] for i in range(0, 784)]).T
X = X / 255
```

```python
X.max()
```

```python
y = utils.to_categorical(data.label, 10)
```

```python
model = Sequential()

model.add(Dense(900, input_dim = 784, activation="relu"))
model.add(Dense(1000, activation="relu"))
model.add(Dense(1500, activation="relu"))
model.add(Dense(2000, activation="relu"))
model.add(Dense(10, activation="softmax"))

model.compile(loss="categorical_crossentropy", optimizer="Adam", metrics=["accuracy"])
print(model.summary())

history = model.fit(X, y, batch_size=60, epochs=5, validation_split=0.3, verbose=1)
```

```python
print(history.history.keys())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
history_model = pd.DataFrame([history.history["accuracy"],history.history["val_accuracy"],history.history["loss"],history.history["val_loss"]]).T
history_model.rename(columns={0:"accuracy", 1: "val_accuracy", 2:"loss", 3:"val_loss"})
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(data=history_model)
```

```python
pr_X = model.predict(X)
```

```python
np.argmax(pr_X[0]) == np.argmax(y[0])
```

```python
model_finish = Sequential()

model_finish.add(Dense(900, input_dim = 784, activation="relu"))
model_finish.add(Dense(1000, activation="relu"))
model_finish.add(Dense(1500, activation="relu"))
model_finish.add(Dense(2000, activation="relu"))
model_finish.add(Dense(10, activation="softmax"))

model_finish.compile(loss="categorical_crossentropy", optimizer="Adam", metrics=["accuracy"])
print(model.summary())

model_finish.fit(X, y, batch_size=60, epochs=2, verbose=1)
```

```python
sample_sub = pd.read_csv('../input/Kannada-MNIST/sample_submission.csv')
dig_mnist = pd.read_csv('../input/Kannada-MNIST/Dig-MNIST.csv')
test = pd.read_csv('../input/Kannada-MNIST/test.csv')
```
