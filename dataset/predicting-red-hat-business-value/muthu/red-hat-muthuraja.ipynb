We won't be able to impute activity_id. We have two options - drop the rows where it is not present or keep the rows where activity_id is null as a separate group. 

The first option is a bad one because there are multiple rows where activity_id is not present in the test set as well (see below code). If we drop in train set then, we won't be able to predict for test set.

Though the second method might be incorrect, it is the best course of action because a) we don't have domain knowledge and b) the column names are hidden. In this way, we are not biasing the results - if we impute activity id to be something, then we run the risk of biasing the results.

1. Now let's develop Logistic regression (with lasso for reducing the number of variables), Random Forest (for reducing the number of variables), and XGBoost. Let's do different combinations of ensemble of the methods to get better results.

```python
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score

# split data into train and validation sets
seed = 123
val_size = 0.33
new_X_train2, X_val, Y_train2, Y_val = train_test_split(new_X_train, Y_train, test_size=val_size, random_state=seed)
```

```python
# fit model no training data
model = XGBClassifier()
model.fit(new_X_train2, Y_train2)
```

```python
# make predictions for validation data
y_pred = model.predict(X_val)
predictions = [round(value) for value in y_pred]
```

```python
# evaluate predictions
roc_auc = roc_auc_score(Y_val, predictions)
print("Area under ROC Curve: %f" % (roc_auc * 100.0))
```
