```python
import pandas as pd

application_train = pd.read_csv('../input/home-credit-default-risk/application_train.csv')
application_test  = pd.read_csv('../input/home-credit-default-risk/application_test.csv')

previous_application = pd.read_csv('../input/home-credit-default-risk/previous_application.csv')
```

```python
from autox.autox_competition.feature_engineer import FeatureOne2Many

featureOne2Many = FeatureOne2Many()
featureOne2Many.fit(application_train, previous_application, on = ['SK_ID_CURR'], target = 'TARGET')
```

```python
one2many_train = featureOne2Many.transform(application_train)
one2many_test  = featureOne2Many.transform(application_test)
```
