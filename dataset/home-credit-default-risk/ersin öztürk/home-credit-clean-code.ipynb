```python
# Sadece categorik datanın sayısını verir

def bar_plot(variable):
    """
        input: variable ex:"Sex"
        output: bar plot & value count
    """
    
    # get feature
    var= df_application_train[variable]
    
    # count number of categorical variable (value/sample)
    varValue=var.value_counts()
    
    print("{}: \n {}".format(variable,varValue))
    
    # visualize
    plt.figure(figsize=(9,3))
    plt.bar(varValue.index, varValue)
    plt.xticks(varValue.index, varValue.index.values)
    plt.ylabel("Frequency")
    plt.title(variable)
    plt.show()
    
    
```

```python
def bar_plot_values(variable):
    
    # get feature
    var= df_application_train[variable]
    
    # count number of categorical variable (value/sample)
    varValue=var.value_counts()
    print("\n","\n","\n")
    print("{}: \n {}".format(variable,varValue))
```

```python
# özelliğin bulunma oranı ve targeta etkisi
def plot_stats(feature,label_rotation=False,horizontal_layout=True):
    temp = df_application_train[feature].value_counts()
    df1 = pd.DataFrame({feature: temp.index,'Number of contracts': temp.values})

    # Calculate the percentage of target=1 per category value
    cat_perc = df_application_train[[feature, 'TARGET']].groupby([feature],as_index=False).mean()
    cat_perc.sort_values(by='TARGET', ascending=False, inplace=True)
    
    if(horizontal_layout):
        fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,6))
    else:
        fig, (ax1, ax2) = plt.subplots(nrows=2, figsize=(12,14))
    sns.set_color_codes("pastel")
    s = sns.barplot(ax=ax1, x = feature, y="Number of contracts",data=df1)
    if(label_rotation):
        s.set_xticklabels(s.get_xticklabels(),rotation=90)
    
    s = sns.barplot(ax=ax2, x = feature, y='TARGET', order=cat_perc[feature], data=cat_perc)
    if(label_rotation):
        s.set_xticklabels(s.get_xticklabels(),rotation=90)
    plt.ylabel('Percent of target with value 1 [%]', fontsize=10)
    plt.tick_params(axis='both', which='major', labelsize=10)

    plt.show();
```

```python
# yazı olarak özelliğin targeta etkisi
def efect_target(feature):
    
    print(df_application_train[[feature,"TARGET"]].groupby([feature],as_index=False).mean().sort_values(by="TARGET", ascending=False))
    print("\n")
```

# OUTLIAR DETECTION¶


```python
def detect_outliers(df,features):
    outlier_indices = []
    
    for c in features:
        # 1st quartile
        Q1 = np.percentile(df[c],25)
        # 3rd quartile
        Q3 = np.percentile(df[c],75)
        # IQR
        IQR = Q3 - Q1
                        #print(Q1,Q3,IQR)
        # Outlier step
        outlier_step = IQR * 1.5
        # detect outlier and their indeces
        outlier_list_col = df[(df[c] < Q1 - outlier_step) | (df[c] > Q3 + outlier_step)].index
        # store indeces
        outlier_indices.extend(outlier_list_col)
    
    outlier_indices = Counter(outlier_indices)
    multiple_outliers = list(i for i, v in outlier_indices.items() if v > 2)
    
    return multiple_outliers
```

```python
def missing_data(data):
    total = data.isnull().sum().sort_values(ascending = False)
    percent = (data.isnull().sum()/data.isnull().count()*100).sort_values(ascending = False)
    return pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
```

```python
def find_max_3_corelation(feature):
    
    df_feature=pd.DataFrame(df_application_train).corr()[[feature]]
    df_feature_abs=abs(df_feature)
    df_feature_abs.sort_values(by=feature,ascending=False).head(5).index[1:4]
    index_name_of_max_corr=df_feature_abs.sort_values(by=feature,ascending=False).head(5).index[1:4]
    
    
    return index_name_of_max_corr
    
```

```python
def max_corr_result(feature):
    df_feature=pd.DataFrame(df_application_train).corr()[[feature]]
    df_feature_abs=abs(df_feature)
    print(df_feature_abs.sort_values(by=feature,ascending=False).head(5))
    print(df_feature_abs.sort_values(by=feature,ascending=False).head(5).index[1:4])
```

```python
find_max_3_corelation('AMT_ANNUITY')
```

```python
def fill_accordingto_max_3_corelation(feature):
    index_nan = list(df_application_train[feature][df_application_train[feature].isnull()].index)
    for i in index_nan:
        feature_pred = df_application_train[feature][
            ((df_application_train[find_max_3_corelation(feature)[0]] == df_application_train.iloc[i][find_max_3_corelation(feature)[0]]) &
             (df_application_train[find_max_3_corelation(feature)[1]] == df_application_train.iloc[i][find_max_3_corelation(feature)[1]]) &
            (df_application_train[find_max_3_corelation(feature)[2]] == df_application_train.iloc[i][find_max_3_corelation(feature)[2]])
            )].median()
        feature_med = df_application_train[feature].median()
        if not np.isnan(feature_pred):
            df_application_train[feature].iloc[i] = feature_pred
        else:
            df_application_train[feature].iloc[i] = feature_med
    
    
   
```

```python
def find_max_3_corelation_test(feature):
    
    df_feature=pd.DataFrame(df_application_test).corr()[[feature]]
    df_feature_abs=abs(df_feature)
    df_feature_abs.sort_values(by=feature,ascending=False).head(5).index[1:4]
    index_name_of_max_corr=df_feature_abs.sort_values(by=feature,ascending=False).head(5).index[1:4]
    
    
    return index_name_of_max_corr
```

```python
def max_corr_result_test(feature):
    df_feature=pd.DataFrame(df_application_test).corr()[[feature]]
    df_feature_abs=abs(df_feature)
    print(df_feature_abs.sort_values(by=feature,ascending=False).head(5))
    print(df_feature_abs.sort_values(by=feature,ascending=False).head(5).index[1:4])
```

```python
def fill_accordingto_max_3_corelation_test(feature):
    index_nan = list(df_application_test[feature][df_application_test[feature].isnull()].index)
    for i in index_nan:
        feature_pred = df_application_test[feature][
            ((df_application_test[find_max_3_corelation_test(feature)[0]] == df_application_test.iloc[i][find_max_3_corelation_test(feature)[0]]) &
             (df_application_test[find_max_3_corelation_test(feature)[1]] == df_application_test.iloc[i][find_max_3_corelation_test(feature)[1]]) &
            (df_application_test[find_max_3_corelation_test(feature)[2]] == df_application_test.iloc[i][find_max_3_corelation_test(feature)[2]])
            )].median()
        feature_med = df_application_test[feature].median()
        if not np.isnan(feature_pred):
            df_application_test[feature].iloc[i] = feature_pred
        else:
            df_application_test[feature].iloc[i] = feature_med
```

```python
def bar_plot_new(variable):

    
    # get feature
    var= df[variable]
    
    # count number of categorical variable (value/sample)
    varValue=var.value_counts()
    
    print("{}: \n {}".format(variable,varValue))
    
    # visualize
    plt.figure(figsize=(9,3))
    plt.bar(varValue.index, varValue)
    plt.xticks(varValue.index, varValue.index.values)
    plt.ylabel("Frequency")
    plt.title(variable)
    plt.show()
```

```python
bar_plot_new("TARGET")
```

```python
y = df["TARGET"]
X = df.drop(['TARGET'], axis=1)
X = pd.DataFrame(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.30, 
                                                    random_state=42)

rf_model = RandomForestClassifier().fit(X_train, y_train)

y_pred = rf_model.predict(X_test)

print(accuracy_score(y_test, y_pred))
print()
print(classification_report(y_test, y_pred))
print()
print(confusion_matrix(y_test,y_pred))
```
