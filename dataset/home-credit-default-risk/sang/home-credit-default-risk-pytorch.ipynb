```python
import torch
import torch.nn as nn
```

```python
torch_xtrain = torch.tensor(xtrain_smote.values)
torch_xtest = torch.tensor(x_test.values)
torch_ytrain = torch.tensor(ytrain_smote.values)
torch_ytest = torch.tensor(y_test.values)
```

```python
print(torch_xtrain.shape)
print(torch_xtest.shape)
print(torch_ytrain.shape)
print(torch_ytest.shape)
```

```python
torch_ytest = torch_ytest.view(61503, 1)
```

```python
class LogisticRegression(nn.Module):
    def __init__(self, input_size, H1, output_size):
        super(LogisticRegression, self).__init__()
        self.linear = nn.Linear(input_size, H1)
        self.linear2 = nn.Linear(H1, output_size)

    def forward(self, x):
        x = torch.sigmoid(self.linear(x))
        x = torch.sigmoid(self.linear2(x))
        return x
    
    def predict(self, x):
        pred = self.forward(x)
        if pred >= 0.5:
            return 1
        else:
            return 0
```

```python
torch.manual_seed(2)
```

```python
LR = LogisticRegression(86 , 200,  1)
```

```python
y_prd = LR.forward(torch_xtest.float())
```

```python
y_prd
```

```python
criterion = nn.BCELoss()
optimizer = torch.optim.Adam(LR.parameters() , lr = 0.001)
```

```python
loss = criterion(y_prd , torch_ytest.float())
print(loss)
```

```python
epochs = 100
losses = []
for i in range(epochs):
    prd = LR.forward(torch_xtest.float())
    loss = criterion(prd , torch_ytest.float())
    #print("epoch:", i , " loss", loss.item())
    losses.append(loss.item())
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
print(losses[0], losses[-1])
```

```python
plt.plot(range(epochs), losses)
```
