### Team Three Competition Submission

```python
import numpy as np
import pandas as pd

import joblib
```

#### Load data and review features

```python
test = pd.read_csv('../input/home-credit-default-risk/application_test.csv')
test.set_index(['SK_ID_CURR'], inplace=True)
test.shape
```

```python
test = test[['AMT_INCOME_TOTAL', 'AMT_CREDIT', 'DAYS_EMPLOYED', 'NAME_CONTRACT_TYPE', 'CODE_GENDER', 'FLAG_OWN_CAR', 'FLAG_OWN_REALTY',
            'NAME_EDUCATION_TYPE', 'NAME_FAMILY_STATUS', 'NAME_HOUSING_TYPE']]
```

```python
test.head(15)
```

#### Make Predictions

```python
#load saved pipeline and model
preprocessor = joblib.load('../input/wk1default-risk-v2/default_preprocessor (1).joblib')

rf_model = joblib.load('../input/wk1default-risk-v2/rf_default_model (1).joblib')
```

```python
X_test = preprocessor.transform(test)
```

```python
print(X_test.shape)
```

```python
test_prob = rf_model.predict_proba(X_test)
print(test_prob[:10])
```

#### Save submission

```python
submission = pd.read_csv('../input/home-credit-default-risk/sample_submission.csv')
submission.head()
```

```python
submission.TARGET = test_prob[:,1]
submission.head()
```

```python
submission.to_csv('Team3_Week1_submission.csv', index=False, header = True)
```
