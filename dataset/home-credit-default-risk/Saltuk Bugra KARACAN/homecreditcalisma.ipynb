# Previous Application

**previous_application.csv = HomeCredit firmasına daha önce yapılan kredi başvuruları. Her satır bir kredi başvurusu**

- previous_application.csv	SK_ID_PREV 	Home Credit kredisi ile ilişkili önceki kredilerin ID’si

- previous_application.csv	SK_ID_CURR	Kredinin ID numarası

- previous_application.csv	NAME_CONTRACT_TYPE	Başvurunun sözleşme ürünü türü (Nakit kredi, tüketici kredisi [POS], ...)

- previous_application.csv	AMT_ANNUITY	Bir önceki başvurunun rantı. ödenecek taksit tutarı olabilir?

- previous_application.csv	AMT_APPLICATION	Müşteri ne kadar kredi istedi? Başvurduğu kredi miktarı 

- previous_application.csv	AMT_CREDIT	Bir önceki başvurunun en son kredi miktarı, Bankanın verebileceği kredi miktarı yani kişinin bankadaki kredi limiti

- previous_application.csv	AMT_DOWN_PAYMENT	Bir önceki başvuruda peşinat Verdiği peşinat miktarı

- previous_application.csv	AMT_GOODS_PRICE	Bir önceki başvuruda müşterinin talep ettiği malın fiyatı (eğer varsa) Müşterinin almak istediği ürünün fiyatı nedir?

- previous_application.csv	WEEKDAY_APPR_PROCESS_START	Müşteri, başvurusunu haftanın hangi günü yapmıştır?

- previous_application.csv	HOUR_APPR_PROCESS_START	Müşteri, krediye yaklaşık saat kaçta başvurdu?

- previous_application.csv	FLAG_LAST_APPL_PER_CONTRACT	Bir önceki sözleşme için son başvuru ise işaretle. Bazen müşteri ya da memurumuzun hatası ile tek bir sözleşme için daha fazla başvuru olabilir. ?

- previous_application.csv	NFLAG_LAST_APPL_IN_DAY	Müşterinin başvurusu, o gün yaptığı son başvuru mu? Bazen müşteriler gün içinde birden fazla başvuru yapıyor. Nadiren veritabanımızda tek başvuru iki kere olup hataya sebebiyet verebiliyor ?

- previous_application.csv	NFLAG_MICRO_CASH	Mikro finans kredisi (Evet, Hayır)

- previous_application.csv	RATE_DOWN_PAYMENT	Önceki kredi üzerinde normalleştirilmiş peşinat oranı

- previous_application.csv	RATE_INTEREST_PRIMARY	Önceki kredi normalleştirilmiş faiz oranı

- previous_application.csv	RATE_INTEREST_PRIVILEGED	Önceki kredi normalleştirilmiş faiz oranı  ayrıcalık mı?

- previous_application.csv	NAME_CASH_LOAN_PURPOSE	Nakit kredinin amacı

- previous_application.csv	NAME_CONTRACT_STATUS	Kredinin sözleşme durumu (aktif imzalı, ...)

- previous_application.csv	DAYS_DECISION	Mevcut başvuru ile ilgili olarak, önceki başvuruya ne zaman kara verildi?

- previous_application.csv	NAME_PAYMENT_TYPE	Müşterinin kredi için seçtiği ödeme yöntemi

- previous_application.csv	CODE_REJECT_REASON	Başvuru neden reddedildi?

- previous_application.csv	NAME_TYPE_SUITE	Müşteri krediye başvururken yanında kim vardı?

- previous_application.csv	NAME_CLIENT_TYPE	Müşteri krediye başvururken yeni mi yoksa eski müşteri miydi?

- previous_application.csv	NAME_GOODS_CATEGORY	Müşteri krediye başvururken ne tarz ürünler için başvurdu?

- previous_application.csv	NAME_PORTFOLIO	Başvuru ne için? (Nakit, POS, araba, ...)

- previous_application.csv	NAME_PRODUCT_TYPE	Başvuru çapraz satış mı yoksa walk-in mi?

- previous_application.csv	CHANNEL_TYPE	Hangi kanalı kullanarak müşteriyi elde ettik?

- previous_application.csv	SELLERPLACE_AREA	Önceki başvurunun satıcı yeri satış alanı

- previous_application.csv	NAME_SELLER_INDUSTRY	Satıcının endüstrisi

- previous_application.csv	CNT_PAYMENT	Term of previous credit at application of the previous application (çeviremedim)

- previous_application.csv	NAME_YIELD_GROUP	Bir önceki krediye göre gruplandırılmış faiz oranı

- previous_application.csv	PRODUCT_COMBINATION	Kredinin ayrıntılı ürün kombinasyonu

- previous_application.csv	DAYS_FIRST_DRAWING	Mevcut başvurunun başvuru tarihine göre, bir önceki başvurunun ilk ödemesi yapıldığı zaman

- previous_application.csv	DAYS_FIRST_DUE	Mevcut başvurunun başvuru tarihine göre, bir önceki başvurunun ilk vadesi ne zaman olmalıydı?

- previous_application.csv	DAYS_LAST_DUE_1ST_VERSION	Mevcut başvurunun başvuru tarihine göre, bir önceki başvurunun ilk vadesi ne zamandı?

- previous_application.csv	DAYS_LAST_DUE	Mevcut başvurunun başvuru tarihine göre, bir önceki başvurunun son vade tarihi ne zamandı?

- previous_application.csv	DAYS_TERMINATION	Mevcut başvurunun başvuru tarihine göre, bir önceki başvurunun beklenen bitişi ne zamandı?

- previous_application.csv	NFLAG_INSURED_ON_APPROVAL	Müşteri bir önceki başvurunda sigorta talep etti mi?

```python
def installments_payments(num_rows = None, nan_as_category = True):
    ins = pd.read_csv('../input/installments_payments.csv', nrows = num_rows)
    
    #ins, cat_cols = one_hot_encoder(ins, nan_as_category= True)
    
    # Percentage and difference paid in each installment (amount paid and installment value)
    ins['PAYMENT_PERC'] = ins['AMT_PAYMENT'] / ins['AMT_INSTALMENT']
    ins['PAYMENT_DIFF'] = ins['AMT_INSTALMENT'] - ins['AMT_PAYMENT']
    
    # Days past due and days before due (no negative values)
    ins['DPD'] = ins['DAYS_ENTRY_PAYMENT'] - ins['DAYS_INSTALMENT']
    ins['DBD'] = ins['DAYS_INSTALMENT'] - ins['DAYS_ENTRY_PAYMENT']
    
    
    
    ins['DPD'] = ins['DPD'].apply(lambda x: x if x > 0 else 0) sor
    ins['DBD'] = ins['DBD'].apply(lambda x: x if x > 0 else 0)
    
    
    
    # Features: Perform aggregations
    aggregations = {
        'NUM_INSTALMENT_VERSION': ['nunique'],
        'DPD': ['max', 'mean', 'sum'],
        'DBD': ['max', 'mean', 'sum'],
        'PAYMENT_PERC': ['max', 'mean', 'sum', 'var'],
        'PAYMENT_DIFF': ['max', 'mean', 'sum', 'var'],
        'AMT_INSTALMENT': ['max', 'mean', 'sum'],
        'AMT_PAYMENT': ['min', 'max', 'mean', 'sum'],
        'DAYS_ENTRY_PAYMENT': ['max', 'mean', 'sum']
    }
    
    for cat in cat_cols:
        aggregations[cat] = ['mean']
    ins_agg = ins.groupby('SK_ID_CURR').agg(aggregations)
    ins_agg.columns = pd.Index(['INSTAL_' + e[0] + "_" + e[1].upper() for e in ins_agg.columns.tolist()])
    # Count installments accounts
    ins_agg['INSTAL_COUNT'] = ins.groupby('SK_ID_CURR').size()
    del ins
    gc.collect()
    return ins_agg
```
