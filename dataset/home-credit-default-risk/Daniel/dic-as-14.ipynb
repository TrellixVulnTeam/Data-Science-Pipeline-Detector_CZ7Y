```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('../input/home-credit-default-risk/application_train.csv')
df.head()
```

```python
# Select 2 variables as in the pre-assignment
X = df[['DAYS_BIRTH', 'REGION_RATING_CLIENT_W_CITY']].values
y = df['TARGET'].values
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.75, test_size=0.25)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# Normalized the data
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Choose LogisticRegression with C = 0.0001 for estimating results.
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression(C = 0.0001)
lg.fit(X_train_scaled, y_train)
lg_predict = lg.predict_proba(X_test_scaled)
lg_predict
```

```python
# Predict with application_train.csv
df_test = pd.read_csv('../input/home-credit-default-risk/application_test.csv')
df_test
```

```python
X_test_file = df_test[['DAYS_BIRTH', 'REGION_RATING_CLIENT_W_CITY']].values
```

```python
# Normalized the data
X_test_file_scaled = scaler.transform(X_test_file)
X_test_file_scaled
```

```python
lg_predict = lg.predict_proba(X_test_file_scaled)
lg_predict
```

```python
pred = pd.Series(lg_predict[:, -1], name='TARGET')
final = pd.concat([df_test['SK_ID_CURR'], pred], axis=1)
final
```

```python
import os
final.to_csv(os.path.join('','oof_all_01.csv'), index=False)
```
