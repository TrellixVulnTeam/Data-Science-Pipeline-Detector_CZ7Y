# 各変数の説明
## application_train.csv
- SK_ID_CURR  
借りた人のID
- TARGET  
返済できたか否か(返済できなかった場合が1)(目的変数)
- NAME_CONTRADCT_TYPE  
キャッシュローンなのかリボルビングローンなのか(一般的な返済か毎月一定額の支払いにならすか)
- CODE_GENDER  
F(女性)/M(男性)
- FLAG_OWN_CAR  
Y(自動車持っている)/N(持っていない)
- FLAG_OWN_REALTY  
Y(家やアパートを保持している)/N(保持していない)
- CNT_CHILDREN  
子供が何人いるか
- AMT_INCOME_TOTAL  
収入(期間や単位がよくわからない)
- AMT_CREDIT  
ローンの支払額
- AMT_ANNUITY  
一年あたりのローンの支払額
- AMT_GOODS_PRICE  
家の価格(だと思う)(ローンをした目的の商品の価格)
- NAME_TYPE_SUITE  
ローンを申し込んだ時に一緒に来た人
- NAME_INCOME_TYPE  
収入源の種類(労働/年金/使用人として勤めているとか)
- NAME_EDUCATION_TYPE  
最終学歴
- NAME_FAMILY_STATUS  
婚姻状況(独身/既婚/未亡人など)
- NAME_HOUSING_TYPE  
家を借りる目的
- REGION_POPULATION_RELATIVE  
顧客が住んでいる場所の人口(正規化済み)
- DAYS_BIRTH  
申込み時点で生まれてから何日たっているか(負の数になっている)
- DAYS_EMPLOYED  
申込み時点で現在の職業を何日やっているか(負の数になっている)
- DAYS_REGISTERATION  
顧客が登録を変えてから何日たっているか(負の数になっている)(よくわからない)
- DAYS_ID_PUBLISH  
ローンを申請した際に使った証明書を何日前に更新していたか(負の数になっている)
- OWN_CAR_AGE  
車を所有してから何年たつか(所有していない場合はnull)
- FLAG_MOBIL  
携帯電話番号を提供したかどうか(した場合1)
- FLAG_EMP_PHONE  
仕事の電話番号を提供したかどうか(した場合1)(多分こうだと思う)
- FLAG_WORK_PHONE  
会社の電話番号を提供したかどうか(した場合は1)
- FLAG_CONT_MOBILE  
携帯電話が通じるか(通じる場合は1)
- FLAG_PHONE  
自宅の電話番号を提供したかどうか(した場合は1)
- FLAG_EMAIL  
メールアドレスを提供したかどうか(した場合は1)
- OCCUPATION_TYPE  
どのような職業についているか(ついてない場合はnull)(NAME_INCOME_TYPEとの多重共線性に注意)
- CNT_FAM_MEMBERS  
何人家族か
- REGION_RATING_CLIENT  
今住んでいる地域の評価(1,2,3)
- REGION_RATING_CLIENT_W_CITY  
今住んでいる都市の評価(1,2,3)
- WEEKDAY_APPR_PROCESS_START  
ローンを申し込んだ曜日
- HOUR_APPR_PROCESS_START  
何時にローンを申し込んだか(整数)
- REG_REGION_NOT_LIVE_REGION  
今住んでいる地域は本籍地と同じか(異なると1,同じなら0)
- REG_REGION_NOT_WORK_REGION  
今働いている地域は本籍地と同じか(異なると1,同じなら0)
- LIVE_REGION_NOT_WORK_REGION  
今住んでいる地域は働いている地域と同じか(異なると1,同じなら0)
- REG_CITY_NOT_LIVE_CITY  
今住んでいる都市は本籍地と同じか(異なると1,同じなら0)
- REG_CITY_NOT_WORK_CITY  
今働いている都市は本籍地と同じか(異なると1,同じなら0)
- LIVE_CITY_NOT_WORK_CITY  
今住んでいる都市は働いている地域と同じか(異なると1,同じなら0)
- ORGANIZATION_TYPE  
働いている組織タイプ(政府とかレストランのような具体的なものから、Type3のような抽象的なものまである)
- EXT_SOURCE_1  
- EXT_SOURCE_2  
- EXT_SOURCE_3  
外部のデータソースからの顧客評価(0~1)
- APARTMENTS_AVG/MODE/MEDI  
住んでいる建物におけるアパートの大きさの平均/最頻値/中央値(0~1)
- BASEMENTAREA_AVG/MODE/MEDI  
基礎部分の大きさ
- YEARS_BEGINEXPLUATATION_AVG/MODE/MEDI  
- YEARS_BUILD_AVG/MODE/MEDI  
建物の年数
- COMMONAREA_AVG/MODE/MEDI  
共用部分
- ELEVATORS_AVG/MODE/MEDI  
エレベーターの数
- ENTRANCES_AVG/MODE/MEDI  
入り口の数
- FLOORSMAX_AVG/MODE/MEDI  
一番上の階は何階か
- FLOORSMIN_AVG/MODE/MEDI  
一番下の階は何階か
- LANDAREA_AVG/MODE/MEDI  
土地の広さ
- LIVINGAPARTMENTS_AVG/MODE/MEDI  
- LIVINGAREA_AVG/MODE/MEDI  
居住部分の大きさ
- NONLIVINGAPARTMENTS_AVG/MODE/MEDI  
- NONLIVINGAREA_AVG/MODE/MEDI  
非居住部分の大きさ
- FONDKAPREMONT_MODE  
- HOUSETYPE_MODE  
ほとんどblock of flatsかnull
- TOTALAREA_MODE  
- WALLSMATERIAL_MODE  
panelとかStone, brickとか
- EMERGENCYSTATE_MODE  
ほとんどnoかnull
(AVG/MODE/MEDIで表される建物情報はnullも多い(50%とか))
- OBS_30_CNT_SOCIAL_CIRCLE  
申し込み時に、返済期日から30日経過している人が何人いるか
- DEF_30_CNT_SOCIAL_CIRCLE  
申し込み時に、期日から30日経過してデフォルトした人数が何人か
- OBS_60_CNT_SOCIAL_CIRCLE  
申し込み時に、返済期日から60日経過している人が何人いるか
- DEF_60_CNT_SOCIAL_CIRCLE  
申し込み時に、期日から60日経過してデフォルトした人数が何人か
- DAYS_LAST_PHONE_CHANGE  
以前携帯を変えてから何日たつか
- FLAG_DOCUMENT_i(iは2から21の整数)
書類iを提出したかどうか(したら1)(書類によってほぼ100%提出のものから、30%くらい提出してるものまで様々)
- AMT_REQ_CREDIT_BUREAU_HOUR/DAY/WEEK/MON/QRT/YEAR  
申し込み前１時間/1日/1週間/1ヶ月/1四半期/1年に顧客が問い合わせた回数(それぞれ1日の場合だと1時間の数字は除かれている)  
  
地域：本郷  
都市：文京区  
のような感じ(大きさ的には地域<都市)
