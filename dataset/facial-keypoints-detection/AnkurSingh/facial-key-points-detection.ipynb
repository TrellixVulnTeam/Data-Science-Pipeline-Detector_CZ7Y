```python
df = pd.read_csv('../input/training/training.csv')
df.dropna(inplace=True)
```

```python
df.head(1)
```

```python
df.shape
```

```python
images = df['Image'] 
features = df.drop(['Image'], axis=1)
```

```python
## preparing Images
Images_ = np.zeros((images.shape[0],96*96))
for i, img in enumerate(images):
    Images_[i] = [float(x) for x in img.split(' ')]
    
Images_ = Images_/255.0
Images_ = Images_.reshape(-1,96,96,1)
```

```python
import matplotlib.pyplot as plt
```

```python
x0 = (features).iloc[5,0::2].values
y0 = (features).iloc[5,1::2].values

plt.imshow(Images_[5].reshape(96,96))
plt.scatter(x0, y0, c='red',s = 5)
```

```python
from keras.models import Sequential
from keras.layers import Dense, Conv2D, Flatten, Dropout
from keras.optimizers import RMSprop, Adam
from keras.callbacks import EarlyStopping
```

```python
## building the model
model = Sequential()
model.add(Conv2D(32, (3,3), activation='relu', input_shape=(96,96,1)))
model.add(Conv2D(32, (3,3), activation='relu'))
model.add(Dropout(0.2))

model.add(Conv2D(64, (3,3), activation='relu'))
model.add(Conv2D(64, (3,3), activation='relu'))
model.add(Dropout(0.2))

model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(30, activation='relu'))

model.compile(optimizer=RMSprop(lr=0.01, decay=0.000005),
              loss = 'mse',
              metrics=['mae'])
```

```python
## training the model
model.fit(Images_, features.values, epochs=25, batch_size=16)
```

```python
test_df = pd.read_csv('../input/test/test.csv')
test_df.head()
```

```python
## preparing test data
test_ = np.zeros((test_df['Image'].shape[0],96*96))
for i, img in enumerate(test_df['Image']):
    test_[i] = [float(x) for x in img.split(' ')]
```

```python
## making predictions
predicted = model.predict(test_.reshape(-1,96,96,1))
```
