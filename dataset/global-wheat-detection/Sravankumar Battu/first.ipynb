```python
#importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
```

```python
train_filename_list = []

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        train_filename_list.append(os.path.join(dirname, filename))

train_filename_list = train_filename_list[2:]
```

```python
#importing the train_dataset
train_data = pd.read_csv("/kaggle/input/global-wheat-detection/train.csv")

path = '/kaggle/input/global-wheat-detection/train/'
```

```python
train_data.head()
```

```python
#Getting image ids

train_img_ids = train_data["image_id"].unique()
train_img_ids
```

```python
def show_img(img_id):
    im = cv2.cvtColor(cv2.imread(path + train_img_ids[img_id] + '.jpg'), cv2.COLOR_BGR2RGB)
    

    boundboxes = train_data[train_data.image_id==train_img_ids[img_id]].bbox.tolist()
    
    L = im.copy()
    for i in range(len(boundboxes)):
        x = int(str(boundboxes[i][1:-1]).split(',')[0][:-2])
        y = int(str(boundboxes[i][1:-1]).split(',')[1][1:-2])
        xw = x + int(str(boundboxes[i][1:-1]).split(',')[2][1:-2])
        yh = y + int(str(boundboxes[i][1:-1]).split(',')[3][1:-2])
        cv2.rectangle(L,(x,y),(xw,yh),(180,190,0),4)
    plt.imshow(L);
    title = "image id : " + str(img_id)
    plt.title(title)
```

```python

pictures = 51
columns = 5
rows = pictures // columns + 1 if pictures % columns else pictures // columns
```

```python
plt.figure(figsize=(30,60))
for i in range(1,pictures+1):
    plt.subplot(rows,columns,i)
    plt.grid(False)
    show_img(i)
```
