```python
from glob import glob
import pickle
from tqdm.notebook import tqdm
import pydicom as dicom
import tensorflow as tf
import numpy as np
import pandas as pd
import cv2
import os
from tensorflow.keras.preprocessing.image import img_to_array, load_img
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
```

```python
def show_top_images_with_lables(images, lables, plot_random=True):
  columns = 4
  rows = 5
  fig=plt.figure(figsize=(12, 17))
  for i in range(1, columns*rows +1):
    index = None
    if plot_random == True:
      index = randint(0, (len(lables)-1))
    else:
      index = i
    fig.add_subplot(rows, columns, i).set_title(lables[index-1])
    plt.imshow(images[index-1], cmap='gray')
  return plt.show()
```

```python
train_dcm_path = '../input/rsna-pneumonia-detection-challenge/stage_2_train_images/*'
test_dcm_path = '../input/rsna-pneumonia-detection-challenge/stage_2_test_images/*'
train_class_csv_path = '../input/rsna-pneumonia-detection-challenge/stage_2_detailed_class_info.csv'
train_lable_csv_path = '../input/rsna-pneumonia-detection-challenge/stage_2_train_labels.csv'
```

```python
train_lables_df = pd.read_csv(train_lable_csv_path)
train_lables_df.head()
```

```python
training_lable_class_df = pd.read_csv(train_class_csv_path)
training_lable_class_df.head()
```

```python
train_dcm_glop = glob(train_dcm_path)
test_dcm_glop = glob(test_dcm_path)
len(train_dcm_glop), len(test_dcm_glop)
```

```python
image_size = (224,224)
```

```python
training_image_data = []
training_image_names = []
print("Reading Images....")
for train_dcm_path in train_dcm_glop:
    with tf.device('/device:GPU:0'):
        file_name = os.path.split(train_dcm_path)[1]
        ds = dicom.dcmread(train_dcm_path)
        image = np.array(ds.pixel_array)
        image = cv2.resize(image, image_size)
        training_image_data.append(image)
        training_image_names.append(file_name)
print("Completed.....")
```

```python
len(training_image_names), len(training_image_data)
```

```python
training_image_data[0].shape
```

```python
file_name = [i+1 for i in range(0,20)]
len(file_name)
```

```python
show_top_images_with_lables(training_image_data, training_image_names, False)
```

```python
training_image_data = np.array(training_image_data)
training_image_names = np.array(training_image_names)
```

```python
training_image_data.shape
```

```python
training_image_names.shape
```

```python
testing_image_data = []
testing_image_names = []
print("Reading Images....")
for test_dcm_path in test_dcm_glop:
    with tf.device('/device:GPU:0'):
        file_name = os.path.split(test_dcm_path)[1]
        ds = dicom.dcmread(test_dcm_path)
        image = np.array(ds.pixel_array)
        image = cv2.resize(image, image_size)
        testing_image_data.append(image)
        testing_image_names.append(file_name)
print("Completed.....")
testing_image_data = np.array(testing_image_data)
testing_image_names = np.array(testing_image_names)
testing_image_data.shape, testing_image_names.shape
```

```python
show_top_images_with_lables(testing_image_data, testing_image_names, False)
```

```python
dataset_dict = {
    "TrainingImagesData": training_image_data,
    "TrainingImageNames": training_image_names,
    "TestingImageData": testing_image_data,
    "TestImageNames": testing_image_names
}

data_pickle_file = open('RSNA-Pneumonia-Detection-Chalenge.pkl', 'ab')
pickle.dump(dataset_dict, data_pickle_file)
data_pickle_file.close() 
```
