```python
import pydicom
import os
from PIL import Image

import torch
import torch.nn as nn
from torch.optim import Adam
from torch.utils.data import DataLoader
from torchvision import transforms , models
from torchvision.datasets import ImageFolder
from torch.autograd import Variable
```

```python
train = os.listdir("/kaggle/input/rsna-pneumonia-detection-challenge/stage_2_train_images")
```

```python
df = pd.read_csv("/kaggle/input/rsna-pneumonia-detection-challenge/stage_2_train_labels.csv")
```

```python
os.mkdir("./train")
os.mkdir("./train/have")
os.mkdir("./train/have_not")
for i in train:
  ds = pydicom.dcmread(r"/kaggle/input/rsna-pneumonia-detection-challenge/stage_2_train_images/" + i)
  name = i.split(".")[0]
  img = Image.fromarray(ds.pixel_array)
  con = df[df.patientId == name].values[0][5]
  if con == 0:
    img.save("./train/have/" + name + ".jpeg")
  else:
    img.save("./train/have_not/" + name + ".jpeg")
```
