```python
import pydicom as dicom
import os
import cv2
import PIL # optional
# make it True if you want in PNG format
PNG = False
# Specify the .dcm folder path
# folder_path = "stage_1_test_images"
folder_path = '/kaggle/input/rsna-pneumonia-detection-challenge/stage_2_train_images/'
# Specify the output jpg/png folder path
jpg_folder_path = "/kaggle/working/dataset"
images_path = os.listdir(folder_path)
for n, image in enumerate(images_path):
    ds = dicom.dcmread(os.path.join(folder_path, image))
    pixel_array_numpy = ds.pixel_array
    
    # resize image
    # can be (256, 256)
    resized = cv2.resize(pixel_array_numpy, (256, 256), interpolation = cv2.INTER_AREA)
    
    if PNG == False:
        image = image.replace('.dcm', '.jpg')
    else:
        image = image.replace('.dcm', '.png')
    cv2.imwrite(os.path.join(jpg_folder_path, image), resized)
    if n % 2000 == 0:
        print('{} image converted'.format(n))
```
