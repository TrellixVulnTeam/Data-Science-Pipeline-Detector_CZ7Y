```python
import pandas as pd
```

```python
sub0 = pd.read_csv('../input/lb4522/submission.csv')
sub1 = pd.read_csv('../input/lb4555/submission.csv')
sub2 = pd.read_csv('../input/lb4722/submission.csv')
```

```python
print(sub0.head())
```

```python
sub0 = sub0.set_index('site_path_timestamp')
sub1 = sub1.set_index('site_path_timestamp')
sub2 = sub2.set_index('site_path_timestamp')
```

```python
sub0['x'] = 0.4*sub0['x'].values + 0.4*sub1.loc[sub0.index, 'x'].values + 0.2*sub2.loc[sub0.index, 'x'].values
sub0['y'] = 0.4*sub0['y'].values + 0.4*sub1.loc[sub0.index,'y'].values + 0.2*sub2.loc[sub0.index,'y'].values
```

```python
sub0 = sub0.reset_index(drop=False)
sub0.to_csv('submission.csv', index=False)
```
