<div class="alert alert-block alert-info", style="font-size:110%">
    <h2><u> Floors Mapping ( Removing Outlier )</u>üíä</h2>

 <p>The competition let us the charge of mapping the floor distribution
    </p>

<h3>Competition Quote</h3>
    <p>
IMPORTANT: The integer floor used in the submission must be mapped from the char/int floors used in the dataset. The mapping is as follows:<br>

* F1, 1F = 0
* F2, 2F  =1
* etc.
* B1, 1B  -1
* B2, 2B  -2
        
There are other floor names in the training data, e.g., LG2, LM, etc., which you may decide to use for training, but none of these non-standard floors are found in the test set.    

</p>   
 
 <h3>My proposition of mapping </h3>
    
* Not important floor (outlier): 'LG1', 'BM', 'BF', 'LG2', 'M', 'G', 'B', 'LM', 'P1', 'P2'

* F1 = 1F = L1 = 0 
* F2 = 2F = L2 = 1
* F3 = L3 = 3F = 2
* F4 = 4F = L4 = 3
* F5 = 5F = L5 = 4
* F6 = 6F = L6 = 5
* F7 = 7F = L7 = 6
* 8F = F8 = L8 = 7
* 9F = F9 = L9 = 8
* L10 = F10 = 9
* L11 = 10
* B1 = -1
* B2 =-2
* B3 = -3
    
Let's discuss about that mapping üó£Ô∏è
    
Nevertheless we can see that more people are in quite in groundfloor 

Why the -1 is  empty ?

```python
info2 = info[info['Floor'].isin(['LG1', 'BM', 'BF', 'LG2', 'M', 'G', 'B', 'LM', 'P1', 'P2']) == False]


mapping = {}

mapping[0] = ["F1", "1F", "L1"]
mapping[1] = ["F2", "2F", "L2"]
mapping[2] = ['F3', 'L3', "3F"]
mapping[3] = ['F4', "4F", 'L4']
mapping[4] = ['F5', "5F", "L5"]
mapping[5] = ["F6", "6F", 'L6']
mapping[6] = ["F7", "7F", 'L7']
mapping[7] = ["8F", "F8", 'L8']
mapping[8] = ['9F', 'F9', 'L9']
mapping[9]= ['L10', 'F10']
mapping[10] = ['L11']
mapping[-1] = ["B1"]
mapping[-2] = ["B2"]
mapping[-3] = ["B3"]

for i in mapping.keys():
    for j in mapping.get(i):
        info2 = info2.replace(j,i)

info2
```

```python
fig = go.Figure()

fig.add_trace(go.Indicator(
    title='Type of Floor',
    mode = "number",
    value = len(info2['Floor'].unique()),
    domain = {'row': 0, 'column': 0}))

fig.add_trace(go.Indicator(
    title = "Number of Devices",
    mode = "number",
    value = len(info2['Mobile'].unique()),
    domain = {'row': 0, 'column': 1}))

fig.add_trace(go.Indicator(
    title= {"text": "Highest Populated Floor <br><span style='font-size:0.7em;color:gray'>ID: {}</span>".format(info2.groupby('Floor').Mobile.count().idxmax())},
    mode = "number",
    value = max(info2.groupby('Floor').Mobile.count()),
    domain = {'row': 1, 'column': 0}))

fig.add_trace(go.Indicator(
    title= {"text": "Fewest populated Floor <br><span style='font-size:0.7em;color:gray'>ID: {}</span>".format(info2.groupby('Floor').Mobile.count().idxmin())},
    mode = "number",
    value = min(info2.groupby('Floor').Mobile.count()),
    domain = {'row': 1, 'column': 1}))

fig.update_layout(
    grid = {'rows': 2, 'columns': 2, 'pattern': "independent"},
)
```

```python
# Graph vizualisation

fig = go.Figure([go.Bar(x=info2.groupby('Floor').Mobile.count().sort_values(ascending=False).index,
                        y=info2.groupby('Floor').Mobile.count().sort_values(ascending=False).values
                       )
                ]
               )
fig.update_layout(height=600,
    xaxis_tickangle=-90,
    title='Floor by population',
    xaxis_tickfont_size=10,
    yaxis=dict(
        title='Number of devices',
        titlefont_size=16,
        tickfont_size=14,
    ),
    xaxis=dict(
        title='Floor',
        titlefont_size=16,
        tickfont_size=14,
    )
)
fig.show()
```
