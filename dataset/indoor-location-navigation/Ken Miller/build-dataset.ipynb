```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os

columns = ['mag_x','mag_y', 'mag_z','mag_u_x','mag_u_y', 'mag_u_z', 'tx', 'ty']
train = pd.DataFrame(columns=columns)

for dirname, _, filenames in os.walk('/kaggle/input/indoor-location-navigation/train/'):
    for filename in filenames:
        #print(os.path.join(dirname, filename))
        df = pd.read_csv(os.path.join(dirname, filename), header = None)
        df['new']=df[0].str.replace('\t',',')
        df.drop(df.columns[0],axis=1,inplace=True)
        #waypoint is row 7
        waypoint=df.iloc[7].values
        
        #mag is row 9
        mag_cal=df.iloc[9].values
    
        #TYPE_MAGNETIC_FIELD_UNCALIBRATED
        mag_uncal=df.iloc[12].values

        waypoint.tolist()
        mag_uncal.tolist()
        mag_cal.tolist()

        way=waypoint[0].split(',')
        mag1=mag_cal[0].split(',')
        mag2=mag_uncal[0].split(',')
        if mag2[2].isalnum()==False:
            try:
                to_append=mag1[2],mag1[3],mag1[4],mag2[2],mag2[3],mag2[4],way[2],way[3]
            except IndexError:
                print('index error')
            else:
                df_length = len(train)
                train.loc[df_length] = to_append
                print (mag1[2],mag1[3],mag1[4],mag2[2],mag2[3],mag2[4],way[2],way[3])
                train.to_csv('train_waypoints.csv', index = False)
                
```
