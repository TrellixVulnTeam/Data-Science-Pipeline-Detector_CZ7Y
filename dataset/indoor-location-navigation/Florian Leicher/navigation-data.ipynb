```python
import re
import os
import pandas as pd
import numpy as np
def extract_trajectory(trajectory_path):
    wifi_logs = []
    waypoint_logs = []
    ssid_logs = set()
    bssid_logs = set()
    with open(trajectory_path) as f:
        floor_name = os.path.basename(os.path.dirname(trajectory_path))
        try:
            level = FLOOR_MAP[floor_name]
        except:
            level = None
        line = f.readline()
        while line:
            # see https://github.com/location-competition/indoor-location-competition-20
            waypoint_match = re.findall(r"(\d{13})\tTYPE_WAYPOINT\t(\d+.\d+)\t(\d+.\d+)", line)
            if waypoint_match:
                waypoint_logs.append((int(waypoint_match[0][0]), float(waypoint_match[0][1]), float(waypoint_match[0][2])))
            wifi_match = re.findall(r"(\d{13})\tTYPE_WIFI\t(\w{40})\t(\w{40})\t(-\d+)\t(\d+)\t\d+", line)
            if wifi_match:
                wifi_logs.append((int(wifi_match[0][0]), wifi_match[0][1], wifi_match[0][2], int(wifi_match[0][3]), int(wifi_match[0][4])))
                ssid_logs.add(wifi_match[0][1])
                bssid_logs.add(wifi_match[0][2])
            line = f.readline()
    
    wifi_pointer, waypoint_pointer = 0, 0
    while wifi_pointer < len(wifi_logs) and waypoint_pointer < len(waypoint_logs) - 1:
#         import pdb; pdb.set_trace()
        t_wifi = int(wifi_logs[wifi_pointer][0])
        t0, x0, y0 = waypoint_logs[waypoint_pointer]
        t1, x1, y1 = waypoint_logs[waypoint_pointer + 1]
        if t_wifi <= t1:
            t_percent = (t_wifi-t0) / (t1-t0)
            x, y = x0 + t_percent * (x1 - x0), y0 + t_percent * (y1 - y0)
            wifi_logs[wifi_pointer] += (x, y, level)
            wifi_pointer += 1
        else:
            waypoint_pointer += 1
    for waypoint_pointer in range(len(waypoint_logs) - 1):
        t0, x0, y0 = waypoint_logs[waypoint_pointer]
        t1, x1, y1 = waypoint_logs[waypoint_pointer + 1]
        waypoint_logs[waypoint_pointer] += (((y1 - y0)**2 + (x1 - x0)**2)**.5 * 1000 / (t1 - t0),)
    
    df_wifi_columns = ("t", "ssid", "bssid", "rssi", "frq", "x", "y", "z")
    try:
        df_wifi = pd.DataFrame(wifi_logs, columns=df_wifi_columns)
    except Exception as e:
#         print("Handled Error: ", e)
        try:
            df_wifi = pd.DataFrame(wifi_logs, columns=df_wifi_columns[:-3])
        except Exception as e:
            print("Handled also: ", e)
            df_wifi = pd.DataFrame(wifi_logs)
    return {
        "waypoint": pd.DataFrame(waypoint_logs, columns=("t", "x", "y", "v")), 
        "wifi": df_wifi,
        "ssid": ssid_logs,
        "bssid": bssid_logs,
    }


def get_building_id(path): 
    with open(path) as f:
        next(f) # first line can be skipped
        return re.findall(r"SiteID:(\w*)", next(f))[0]
```

```python
import json
train_dir = '/kaggle/input/indoor-location-navigation/train'
no_train_files = len([filename for _, _, filenames in os.walk(train_dir) for filename in filenames])
counter = 0
counter_nonskipped = 0

my_building = {}
dimensions = {building_id: {} for building_id in ALL_BUILDINGS}
for dirname, _, filenames in os.walk(train_dir):
    for filename in filenames:
        counter += 1
        trajectory_id = filename.split(".")[0]
        floor_name = os.path.basename(dirname)
        building_id = os.path.basename(os.path.dirname(dirname))
#         if building_id not in RELEVANT_BUILDINGS and building_id != '5d27096c03f801723c31e5e0':
#             continue
        counter_nonskipped += 1
        if floor_name not in dimensions[building_id]:
            dimensions[building_id][floor_name] = (10000, -10000, 10000, -10000)
        if floor_name not in my_building:
            my_building[floor_name] = []
        print(f"TRN{counter}/{no_train_files}", end=" ") # " Bldng: {building_id}, Flr: {floor_name}, Trj: {trajectory_id}")
        trajectory = extract_trajectory(os.path.join(dirname, filename))
        dimensions[building_id][floor_name] = (
            min(dimensions[building_id][floor_name][0], min(trajectory["waypoint"].x)), 
            max(dimensions[building_id][floor_name][1], max(trajectory["waypoint"].x)), 
            min(dimensions[building_id][floor_name][2], min(trajectory["waypoint"].y)), 
            max(dimensions[building_id][floor_name][3], max(trajectory["waypoint"].y)))
        trajectory_dir = f"/kaggle/working/train/{building_id}/{floor_name}/{trajectory_id}/"
        os.makedirs(trajectory_dir, exist_ok=True)
        trajectory["waypoint"].to_csv(f"{trajectory_dir}waypoint.csv", header=True, index=False)
        trajectory["wifi"].to_csv(f"{trajectory_dir}wifi.csv", header=True, index=False)
        my_building[floor_name].append(trajectory["waypoint"])
        if is_interactive() and counter_nonskipped > 10:
            break

    if is_interactive() and counter_nonskipped > 10:
        break
print("done")

with open("./dimension.txt", "w") as f:
    json.dump(dimensions, f)
```
