```python
import pandas as pd
import numpy as np
```

```python
import path
import glob,os
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
base_path = '../input/indoor-navigation-and-location-wifi-features/wifi_features/'
```

```python
submission = pd.read_csv('../input/simple-99-accurate-floor-model/submission.csv')
```

```python
train_files  = sorted(glob.glob('../input/indoor-navigation-and-location-wifi-features/wifi_features/train/*.csv'))
test_files = sorted(glob.glob('../input/indoor-navigation-and-location-wifi-features/wifi_features/test/*.csv'))
```

```python
predictions = list()
for i in range(len(train_files)):
    data = pd.read_csv(train_files[i])
    fea,tar = data.iloc[:,1:-4],data.iloc[:,-4:-2]

    predict = pd.read_csv(test_files[i]).iloc[:,1:-1]
    regressor = KNeighborsRegressor(n_neighbors=1,weights="uniform",algorithm="ball_tree",leaf_size=17)
    regressor.fit(fea,tar)
    predictions.append(regressor.predict(predict))
```

```python
predictions = np.vstack(predictions)
```

```python
submission['x'] = predictions[:,0]
submission['y'] = predictions[:,1]
```

```python
submission.to_csv('submission.csv',index=False)
```
