```python
import pandas as pd
import numpy as np
```

```python
submission_file=pd.read_csv('../input/indoor-location-navigation/sample_submission.csv')
```

```python
submission_file = pd.DataFrame(submission_file['site_path_timestamp'])
submission_file['floor'] = 0
submission_file['x'] = 0
submission_file['y'] = 0
```

```python
second_score = pd.read_csv('../input/indoor-best-submits/submission_2.csv')
first_score = pd.read_csv('../input/indoor-best-submits/submission_1.csv')
fourth_score = pd.read_csv('../input/indoor-best-submits/submission_3.csv')
third_score = pd.read_csv('../input/indoor-best-submits/submission_4.csv')
```

```python
submission_file['first_floor'] = first_score.floor
submission_file['first_x'] = first_score.x
submission_file['first_y'] = first_score.y
```

```python
submission_file['second_floor'] = second_score.floor
submission_file['second_x'] = second_score.x
submission_file['second_y'] = second_score.y
```

```python
submission_file['third_floor'] = third_score.floor
submission_file['third_x'] = third_score.x
submission_file['third_y'] = third_score.y
```

```python
submission_file['fourth_floor'] = fourth_score.floor
submission_file['fourth_x'] = fourth_score.x
submission_file['fourth_y'] = fourth_score.y
```

```python
submission_file['mean_x'] = submission_file[['first_x','second_x','third_x','fourth_x']].mean(axis=1)
submission_file['mean_y'] = submission_file[['first_y','second_y','third_y','fourth_y']].mean(axis=1)
submission_file['frequent_floor'] = submission_file[['first_floor','second_floor','third_floor','fourth_floor']].mode(axis=1)[0]
```

```python
submission_file.sample(10)
```

```python
final_subm = submission_file[['site_path_timestamp','frequent_floor','mean_x','mean_y']]
final_subm.columns = ['site_path_timestamp', 'floor', 'x','y']
final_subm.sample(6)
```

```python
final_subm.to_csv('submission.csv', index=False)
```
