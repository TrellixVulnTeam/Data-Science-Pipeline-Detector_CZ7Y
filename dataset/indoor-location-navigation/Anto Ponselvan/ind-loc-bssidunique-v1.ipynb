# 1) Find Unique BSSID

```python
temp=96
for mall in mall_list:
    mall_train_path = glob.glob('../input/indoor-location-navigation/train/'+mall+'/*/*')
    bssid_all = []
    # print((mall_train_path))
    for path in mall_train_path[96:]:
        mwi_data_wifi = (read_data_file(path)).wifi
        if mwi_data_wifi != []:
            bssid_all += list(mwi_data_wifi[:,2])
    bssid_all_unique = set(bssid_all)
    pd.DataFrame(bssid_all_unique).to_csv(mall+'_bssid.csv')
    print(temp," out of 204")
    temp+=1
```
