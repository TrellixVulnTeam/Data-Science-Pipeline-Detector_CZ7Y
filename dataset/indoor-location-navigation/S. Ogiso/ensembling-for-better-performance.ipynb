```python
import pandas as pd
import numpy as np
```

```python
data_path_list = [
    "../input/indoor-gbm-postprocessing-xy-prediction/submission.csv",
    "../input/wifi-features-with-lightgbm-kfold/submission.csv",
    "../input/simple-99-accurate-floor-model/submission.csv",
    "../input/lstm-by-keras-with-unified-wi-fi-feats/submission.csv",
    "../input/indoor-navigation-snap-to-grid-post-processing/submission_snap_to_grid.csv"
]

position_weights = np.array([
    0.15,
    0.05,
    0,
    0.6,
    0.2
])

floor_weights = np.array([
    0,
    0,
    1,
    0,
    0
])

position_weights = position_weights / np.sum(position_weights) # normalize
floor_weights = floor_weights / np.sum(floor_weights) # normalize
```

```python
df_list = [pd.read_csv(d) for d in data_path_list]
```

```python
sub = pd.read_csv("../input/indoor-location-navigation/sample_submission.csv")

sub.floor = df_list[np.argmax(floor_weights)].floor
sub.x = 0
sub.y = 0

for df, w in zip(df_list, position_weights):
    sub[["x", "y"]] += df[["x", "y"]] * w
    
sub.to_csv("submission.csv", index=False)
```
