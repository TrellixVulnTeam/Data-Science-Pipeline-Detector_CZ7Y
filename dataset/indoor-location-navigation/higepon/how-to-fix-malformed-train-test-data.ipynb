### What is this?
I demonstrate here how to fix malformed lines in train&test dataset.

#### Malformed line example
At line 623 in the train file below. There are **two** data in **one line**, TYPE_BEACON and TYPE_ACCELEROMETER.

```python
!sed '623q;d' /kaggle/input/indoor-location-navigation/train/5cd56ba1e2acfd2d33b603af/B2/5cf75ac5e36a480008125f9d.txt
```

Let's create fixed directory and copy the train file to the directory so that we can compare before & after.

```python
!mkdir fixed
!cp /kaggle/input/indoor-location-navigation/train/5cd56ba1e2acfd2d33b603af/B2/5cf75ac5e36a480008125f9d.txt fixed
```

Now let's run the following sed command to fix the malformed line. Note that the command fix the file inplace because we use -i option.

```python
!cd fixed && find ./ -type f -exec sed -i -e 's/[^^]\([[:digit:]]\{13\}\)\tTYPE/\n\1\tTYPE/g' {} \;
```

Alright, let's use diff command to see before and after. It looks great.

```python
!diff -ur /kaggle/input/indoor-location-navigation/train/5cd56ba1e2acfd2d33b603af/B2/5cf75ac5e36a480008125f9d.txt fixed/5cf75ac5e36a480008125f9d.txt
```
