<h2>Tree based feature importance using <a href="https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html" target="_blank">RandomForest</a> and <a href="https://lightgbm.readthedocs.io/en/latest/" target="_blank">lighgbm</a> </h2>

<h3><b>RandomForest</b></h3>

```python
m = RandomForestRegressor(n_estimators=100)
m.fit(df, y)
```

```python
fi = rf_feat_importance(m,df)
```

```python
def plot_fi(fi):
    return fi.plot('cols', 'imp', 'barh', figsize=(12,7), legend=False)
```

```python
plot_fi(fi[fi.imp>0.005])
```

<h3><b>Lighgbm </b></h3>

```python
model = lgb.LGBMRegressor(n_estimators = 100)
model.fit(df, y)
```

```python
fi = pd.DataFrame()
fi["feature"] = df.columns
fi["importance"] = model.feature_importances_
```

```python
top_imp_f = fi[fi.importance>5]
top_imp_f = top_imp_f.sort_values(by="importance", ascending=False)
        
```

```python
top_imp_f.plot('feature', 'importance', 'barh', figsize=(12,7), legend=False)
```

<h3><b>Eli5</b></h3> 

<h3>eli5 is a way to compute feature importances on any model (logisticRegression,svm ...). <a href="https://eli5.readthedocs.io/en/latest/blackbox/permutation_importance.html" target="_blank">eli5 documentation</a>  + <a href="https://www.kaggle.com/lopuhin/eli5-for-mercari">kernel: eli5 in action</a> </h3>

```python
eli5.show_weights(model, top=40)
```
