```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42, stratify = y)
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
logistic_pred = lr.predict(X_test)
```
