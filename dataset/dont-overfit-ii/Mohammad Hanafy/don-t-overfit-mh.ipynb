```python
df_train = pd.read_csv('/kaggle/input/dont-overfit-ii/train.csv', index_col='id')
df_test = pd.read_csv('/kaggle/input/dont-overfit-ii/test.csv', index_col='id')
```

```python
train_columns = list(df_train.columns)
train_columns.remove('target')

X = df_train[train_columns]
Y = df_train['target']
```

```python
X.info()
```

```python
x_train, x_valid, y_train, y_valid = train_test_split(X,Y, test_size=0.2, stratify=Y, random_state=0)
```

### Different Models

```python
# Logistic Regression

lr_model = LogisticRegression(solver='lbfgs')

lr_model.fit(x_train, y_train)

y_train_predict_lr = lr_model.predict(x_train)
y_valid_predict_lr = lr_model.predict(x_valid)

print('Logistic Regression Train Accuracy = {}'.format(accuracy_score(y_train, y_train_predict_lr)))
print('Logistic Regression Valid Accuracy = {}'.format(accuracy_score(y_valid, y_valid_predict_lr)))
print('Logistic Regression ROC_AUC score = {}'.format(roc_auc_score(y_valid, y_valid_predict_lr)))
```

```python
# Naive Bayes

nb_model = GaussianNB(var_smoothing = 1e-12)

nb_model.fit(x_train, y_train)

y_train_predict_nb = nb_model.predict(x_train)
y_valid_predict_nb = nb_model.predict(x_valid)

print('Naive Bayes Train Accuracy = {}'.format(accuracy_score(y_train, y_train_predict_nb)))
print('Naive Bayes Valid Accuracy = {}'.format(accuracy_score(y_valid, y_valid_predict_nb)))
print('Naive Bayes ROC_AUC score = {}'.format(roc_auc_score(y_valid, y_valid_predict_nb)))
```

```python
# K Nearest Neighbour (KNN)

knn_model = KNeighborsClassifier(n_neighbors=6)

knn_model.fit(x_train, y_train)

y_train_predict_knn = knn_model.predict(x_train)
y_valid_predict_knn = knn_model.predict(x_valid)

print('KNN Train Accuracy = {}'.format(accuracy_score(y_train, y_train_predict_knn)))
print('KNN Valid Accuracy = {}'.format(accuracy_score(y_valid, y_valid_predict_knn)))
print('KNN ROC_AUC score = {}'.format(roc_auc_score(y_valid, y_valid_predict_knn)))
```

```python
# Random Forest 

rf_model = RandomForestClassifier(n_estimators=5, max_depth=3, n_jobs=-3, random_state=123)

rf_model.fit(x_train, y_train)

y_train_predict_rf = rf_model.predict(x_train)
y_valid_predict_rf = rf_model.predict(x_valid)

print('Random Forest Train Accuracy = {}'.format(accuracy_score(y_train, y_train_predict_rf)))
print('Random Forest Valid Accuracy = {}'.format(accuracy_score(y_valid, y_valid_predict_rf)))
print('Random Forest ROC_AUC score = {}'.format(roc_auc_score(y_valid, y_valid_predict_rf)))
```

```python
# Decision Tree 

dt_model = DecisionTreeClassifier(max_depth=2)

dt_model.fit(x_train, y_train)

y_train_predict_dt = dt_model.predict(x_train)
y_valid_predict_dt = dt_model.predict(x_valid)

print('Decision Tree Train Accuracy = {}'.format(accuracy_score(y_train, y_train_predict_dt)))
print('Decision Tree Valid Accuracy = {}'.format(accuracy_score(y_valid, y_valid_predict_dt)))
print('Decision Tree ROC_AUC score = {}'.format(roc_auc_score(y_valid, y_valid_predict_dt)))
```

```python
# Support Vector Machine (SVM)
svm_model = SVC(kernel='sigmoid')

svm_model.fit(x_train, y_train)

y_train_predict_svm = svm_model.predict(x_train)
y_valid_predict_svm = svm_model.predict(x_valid)

print('SVM Train Accuracy = {}'.format(accuracy_score(y_train, y_train_predict_svm)))
print('SVM Valid Accuracy = {}'.format(accuracy_score(y_valid, y_valid_predict_svm)))
print('SVM ROC_AUC score = {}'.format(roc_auc_score(y_valid, y_valid_predict_svm)))
```

```python
# XGBClassifier

xgb_model = XGBClassifier(max_depth=2,gamma=6,eta=0.8,reg_alpha=0.5,reg_lambda=0.5, eval_metric='auc', use_label_encoder=False)

xgb_model.fit(x_train, y_train)

y_train_predict_xgb = xgb_model.predict(x_train)

y_valid_predict_xgb = xgb_model.predict(x_valid)

print('XGB Classifier Train Accuracy = {}'.format(accuracy_score(y_train, y_train_predict_xgb)))
print('XGB Classifier Valid Accuracy = {}'.format(accuracy_score(y_valid, y_valid_predict_xgb)))
print('XGB Classifier ROC_AUC score = {}'.format(roc_auc_score(y_valid, y_valid_predict_xgb)))
```

```python
y_test_predict_xgb = xgb_model.predict(df_test)
```

```python
submission = pd.DataFrame({'id':df_test.index,
                    'target':y_test_predict_xgb})
```
