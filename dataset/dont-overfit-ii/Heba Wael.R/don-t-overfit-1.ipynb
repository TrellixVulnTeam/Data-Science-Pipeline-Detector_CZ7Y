```python
#splitting data (feature, target)
Y = train_data['target']
X = train_data.drop(['id','target'], axis = 1)
test_id = test_data['id']
test = test_data.drop(['id'],axis =1)
```

```python
#scaling data using RobustScaler
from sklearn.preprocessing import RobustScaler
r = RobustScaler()
x = r.fit_transform(X)
testt = r.fit_transform(test)
x= pd.DataFrame(x).astype(float)
testt = pd.DataFrame(testt).astype(float)
```

```python
#appling different models
ridge = linear_model.Ridge()
lasso = linear_model.Lasso()
logregression = linear_model.LogisticRegression(solver='liblinear')
Randomforest_regression = RandomForestRegressor(n_estimators = 10, random_state = 0)
XGB = xgb.XGBRegressor(objective ="reg:logistic", colsample_bytree = 0.9, learning_rate = 0.1,
                max_depth = 5, alpha = 0.1, n_estimators = 20)
clf = DecisionTreeClassifier(max_depth=4)
knn = KNeighborsClassifier(n_neighbors=30)

models = [ridge, lasso, logregression,Randomforest_regression, XGB, clf, knn]
```

```python
def crossval_scores(model):
    scores = cross_val_score(model, x, Y, cv=20, scoring='roc_auc')
    print('Average of scores =', np.mean(scores) ,' \n','std of scores =', np.std(scores),'\n')
    
for model in models:
    print(model)
    crossval_scores(model)
```

```python
#Splitting data (train,valid)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, Y,  test_size=0.25, stratify = Y)
```

```python
#using gridsearch
penalty = ['l1', 'l2']
c = [0.0001, 0.001, 0.01, 0.1, 1, 10, 100, 1000]
class_weight = [{1:0.5, 0:0.5}, {1:0.4, 0:0.6}, {1:0.6, 0:0.4}, {1:0.7, 0:0.3}]
solver = ['liblinear', 'saga']

param_grid = dict(penalty=penalty,
                  C=c,
                  class_weight=class_weight,
                  solver=solver)

grid = GridSearchCV(estimator=logregression, param_grid=param_grid, scoring='roc_auc', verbose=1, n_jobs=-1)
grid_result = grid.fit(x_train, y_train)

print('best Score: ', grid_result.best_score_)
print('best Params: ', grid_result.best_params_)
```
