Dual model :

1rst model is static and reused across tasks:
training / assembling(embeddings?) a DNN model to give it enough prior spacial knowledge and 'tools' to work from (to be detailed)
have a way to hook in and out of the model in a generic way :
portions of the model have to be able to be traversed and combined in arbitrary way, regularization could provide help on main hook points

2nd (smaller) model is trained while evaluating a task (=wiring + minimal intrinsic machinery) :
assembling building blocks from the first model / combining them (creating new connections)

training finishes when the second model manage to find the solutions for the current examples
it is then ready to evaluate the new input

second model is dropped in-between every new task (need to be fast !)
