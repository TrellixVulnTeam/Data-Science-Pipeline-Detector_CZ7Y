```python
import datasets
from transformers import AutoModelForSequenceClassification, AutoTokenizer
import torch
from torch.utils.data import DataLoader
```

```python
# train_data = datasets.load_dataset("csv", data_files=[config['train_file']])
```

```python
train_data = datasets.Dataset.from_pandas(train_df)
```

```python
train_data
```

```python
train_data = train_data.rename_column("score","labels")
train_data = datasets.Dataset.from_dict(train_data[:])
```

```python
train_data = train_data.train_test_split(test_size=0.01)
```

```python
train_data
```

```python
tokenizer = AutoTokenizer.from_pretrained(config['model_name'])

model = AutoModelForSequenceClassification.from_pretrained(config['model_name'], 
                                                         num_labels=1)
```

```python
def encode(x):
    return tokenizer(x['text'], truncation=True, max_length=512)
```

```python
train_data = train_data.map(encode, batched=True)
```

```python
train_data.set_format(type='torch', columns=['input_ids',"attention_mask", "labels"])
```

```python
train_data
```

```python
for a in train_data['train']["input_ids"]:
    if a.shape[0] > 512:
        print(a.shape)
        break
```

```python
from transformers import TrainingArguments
from transformers import Trainer
from transformers import DataCollatorWithPadding

data_collator = DataCollatorWithPadding(tokenizer=tokenizer)
```

```python
trainer_args = TrainingArguments('test_trainer',
                                report_to='none',
                                evaluation_strategy ='epoch',
                                per_device_train_batch_size=16,
                                per_device_eval_batch_size=16,
                                fp16=True
                                )

trainer = Trainer(model=model,
                  args=trainer_args, 
                  train_dataset = train_data['train'],
                  eval_dataset=train_data['test'],
                  data_collator = data_collator,
                  tokenizer=tokenizer)
```

```python
print(trainer_args)
```

```python
trainer.train()
```
