```python
import pandas as pd
from sklearn import model_selection
```

```python
df = pd.read_csv("../input/jigsaw-toxic-severity-rating/validation_data.csv")
df["kfold"] = -1

kf = model_selection.StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
for f, (t_, v_) in enumerate(kf.split(X=df, y=df.worker.values)):
    df.loc[v_, 'kfold'] = f

df.to_csv("jrstc_10folds.csv", index=False)
```
