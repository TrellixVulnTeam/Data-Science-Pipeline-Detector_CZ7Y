```python
sample_sub  = pd.read_csv("/kaggle/input/jigsaw-toxic-severity-rating/sample_submission.csv")
score  = pd.read_csv("/kaggle/input/jigsaw-toxic-severity-rating/comments_to_score.csv")
validate  = pd.read_csv("/kaggle/input/jigsaw-toxic-severity-rating/validation_data.csv")
```

```python
#clean up 
def clean_up(data, col):   
    data[col] = data[col].str.replace('\n', '')  
    data[col] = data[col].str.replace(r'([a-zA-Z]+)([/!?.])([a-zA-Z]+)',r'\1 \2 \3')

    # Replace repeating characters more than 3 times to length of 3
    data[col] = data[col].str.replace(r'([*!?\'])\1\1{2,}',r'\1\1\1')    
    
    # Add space around repeating characters
    data[col] = data[col].str.replace(r'([*!?\']+)',r' \1 ')    
    
    # patterns with repeating characters 
    data[col] = data[col].str.replace(r'([a-zA-Z])\1{2,}\b',r'\1\1')
    data[col] = data[col].str.replace(r'([a-zA-Z])\1\1{2,}\B',r'\1\1\1')
    data[col] = data[col].str.replace(r'[ ]{2,}',' ').str.strip()   
    
    return data  # the function returns the processed value

#clean_up(score,'text')  # display the test function after transformation
score = clean_up(score,'text')  # clear the whole date frame
validate = clean_up(validate,'less_toxic')  # clear the whole date frame
validate = clean_up(validate,'more_toxic')  # clear the whole date frame
```
