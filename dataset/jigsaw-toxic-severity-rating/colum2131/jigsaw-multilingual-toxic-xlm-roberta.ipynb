```python
import os
import torch
from transformers import AutoTokenizer, AutoModel, AutoConfig


DOWNLOADED_MODEL_PATH = 'unitary/multilingual-toxic-xlm-roberta'
DATASET_MODEL_PATH = 'model'

os.mkdir(DATASET_MODEL_PATH)

tokenizer = AutoTokenizer.from_pretrained(DOWNLOADED_MODEL_PATH)
tokenizer.save_pretrained(DATASET_MODEL_PATH)
config = AutoConfig.from_pretrained(DOWNLOADED_MODEL_PATH) 
config.save_pretrained(DATASET_MODEL_PATH)
backbone = AutoModel.from_pretrained(DOWNLOADED_MODEL_PATH, config=config)
backbone.save_pretrained(DATASET_MODEL_PATH)
```
