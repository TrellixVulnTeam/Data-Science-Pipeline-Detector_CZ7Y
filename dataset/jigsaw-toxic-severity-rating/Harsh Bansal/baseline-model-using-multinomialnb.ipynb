```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
df=pd.read_csv("../input/ruddit-jigsaw-dataset-combined-cleaned/toxic_train.csv",usecols=["processed","isOffensive"])
test=pd.read_csv("../input/jigsaw-toxic-severity-rating/comments_to_score.csv")
sample=pd.read_csv("../input/jigsaw-toxic-severity-rating/sample_submission.csv")
df=df.rename(columns={"processed":"text","isOffensive":"score"})
```

```python
min_len = (df['score'] == 1).sum()
df_y0_undersample = df[df['score'] == 0].sample(n=min_len, random_state=201)
df = pd.concat([df[df['score'] == 1], df_y0_undersample]).reset_index(drop=True)
```

```python
vec = TfidfVectorizer()
X = vec.fit_transform(df['text'])
#X = to_sparse_dask_array(X,client)
model=MultinomialNB()
model.fit(X, df['score'])
```

```python
X_test = vec.transform(test['text'])
pred = model.predict_proba(X_test)
p=model.predict(X_test)
```

```python
test['score'] = pred[:, 1]
test[['comment_id', 'score']].to_csv("submission.csv", index=False)

```
