```python
import pandas as pd
import numpy as np
import pandas as pd
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df = pd.read_csv("../input/jigsaw-toxic-comment-classification-challenge/train.csv")
df.head()
```

```python
df['y'] = (df[['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']].sum(axis=1) > 0 ).astype(int)
df = df[['comment_text', 'y']].rename(columns={'comment_text': 'text'})
df.sample(5)
```

```python
min_len = (df['y'] == 1).sum()
df_y0_undersample = df[df['y'] == 0].sample(n=min_len, random_state=201)
df = pd.concat([df[df['y'] == 1], df_y0_undersample])
df['y'].value_counts()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf = TfidfVectorizer()
X = tf.fit_transform(df['text'])
X.shape
```

```python
Y = df['y']
```

```python
model = MultinomialNB()
model.fit(X, df['y'])
```

```python
df_val = pd.read_csv("../input/jigsaw-toxic-severity-rating/validation_data.csv")
X_less_toxic = tf.transform(df_val['less_toxic'])
X_more_toxic = tf.transform(df_val['more_toxic'])

p1 = model.predict(X_less_toxic)
p2 = model.predict(X_more_toxic)
p1
```

```python
df_sub = pd.read_csv("../input/jigsaw-toxic-severity-rating/comments_to_score.csv")
X_test = tf.transform(df_sub['text'])
p3 = model.predict_proba(X_test)
```

```python
df_sub
```

```python
df_sub['score'] = p3[:, 1]
```
