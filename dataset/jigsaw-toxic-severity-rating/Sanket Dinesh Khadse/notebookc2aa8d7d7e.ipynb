```python
data = pd.read_csv(r"/kaggle/input/jigsaw-toxic-severity-rating/comments_to_score.csv")
data
```

```python
from transformers import pipeline
senti_model = pipeline(task = "sentiment-analysis", model = r"./models/") 
```

```python
data["text"] = data["text"].map(lambda i: i[:500]).apply(senti_model).apply(lambda i: i[0]["score"] if i[0]["label"] == "NEGATIVE" else 1-i[0]["score"])
```

```python
data.rename(columns = {'text': 'score'}, inplace = True) 
```

```python
data.to_csv("submission.csv")
```
