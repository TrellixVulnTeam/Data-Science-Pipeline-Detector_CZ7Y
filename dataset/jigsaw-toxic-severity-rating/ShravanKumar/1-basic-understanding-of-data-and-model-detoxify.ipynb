```python
submission_sample_file = "/kaggle/input/jigsaw-toxic-severity-rating/sample_submission.csv"
validations_data_file = "/kaggle/input/jigsaw-toxic-severity-rating/validation_data.csv"
comments_to_score_file = "/kaggle/input/jigsaw-toxic-severity-rating/comments_to_score.csv"
```

```python
df_comments = pd.read_csv(comments_to_score_file)
df_comments.head()
```

```python
df_validations = pd.read_csv(validations_data_file)
df_validations.head()
```

```python
df_submission = pd.read_csv(submission_sample_file)
df_submission.head()
```

```python
input_text = 'Fuck off Wolfkeeper Talk'
results = Detoxify('unbiased').predict([input_text, input_text])
results 
```

```python
validation_list = df_validations[['less_toxic', 'more_toxic']].values.tolist()
```

```python
validation_list[0]
```

```python
unbiased_model = Detoxify('unbiased')
```

```python
unbiased_model.predict(validation_list[0])['toxicity']
```

```python
for i in validation_list[:10]:
    print(unbiased_model.predict(i)['toxicity'])
```

```python
df_comments.head()
```

```python
from collections import defaultdict
D = defaultdict(list)
for idx, row in df_comments.iterrows():
    toxic_score = unbiased_model.predict([row['text'],row['text']])['toxicity'][0]
    D['comment_id'].append(row['comment_id'])
    D['score'].append(toxic_score)
```

```python
submission_df = pd.DataFrame(D)
submission_df
```

```python
submission_df.to_csv("submission.csv", index=False)
```
