```python
for i in categorical :
    dummies=pd.get_dummies(train[i])
    dummies.columns=str(i)+'_'+dummies.columns
    train=pd.concat([train,dummies],axis=1)
    train.drop(i, axis=1, inplace=True)
    dummies=pd.get_dummies(test[i])
    dummies.columns=str(i)+'_'+dummies.columns
    test=pd.concat([test,dummies],axis=1)
    test.drop(i, axis=1, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_data=train.drop(['RefId','IsBadBuy'],axis=1)
train_labels=train['IsBadBuy']
```

```python
X_train,X_test,y_train,y_test=train_test_split(train_data,train_labels, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
KNN=KNeighborsClassifier()
KNN.fit(X_train,y_train)
```

```python
KNN.score(X_test,y_test)
```

```python
predict=KNN.predict(test.drop('RefId',axis=1))
```

```python
submission=pd.DataFrame(data=predict,columns=['IsBadBuy'])
```
