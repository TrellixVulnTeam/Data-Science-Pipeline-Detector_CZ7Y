```python
not_cat=train_df.drop(['RefId','IsBadBuy'],axis=1).columns[train_df.drop(['RefId','IsBadBuy'],axis=1).dtypes!= 'object']
```

```python
for i in not_cat:
    maximum=np.max(train_df[i])
    train_df[i]=train_df[i]/maximum
    maximum_test=np.max(test_df[i])
    test_df[i]=test_df[i]/maximum_test
```

```python
categorical=train_df.drop(['RefId','IsBadBuy'],axis=1).columns[train_df.drop(['RefId','IsBadBuy'],axis=1).dtypes=='object']
```

```python
categorical
```

```python
train_df[categorical[0]]
```

```python
pd.get_dummies(train_df[categorical[0]])
```

```python
for i in categorical:
    dummies=pd.get_dummies(train_df[i])
    dummies.columns=str(i)+'_'+dummies.columns
    train_df=pd.concat([train_df,dummies],axis=1)
    train_df.drop(i,inplace=True,axis=1)
    dummies=pd.get_dummies(test_df[i])
    dummies.columns=str(i)+'_'+dummies.columns
    test_df=pd.concat([test_df,dummies],axis=1)
    test_df.drop(i,inplace=True,axis=1)
```

```python
from sklearn.model_selection import train_test_split
X=train_df.drop(['RefId','IsBadBuy'],axis=1)
y=train_df['IsBadBuy']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42)
```

```python
print(X_train.shape,y_train.shape)
print(X_test.shape,y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
KNN=KNeighborsClassifier(n_neighbors=11)
KNN.fit(X_train,y_train)
```

```python
KNN.score(X_test,y_test)
```

```python
predict=KNN.predict(test_df.drop('RefId',axis=1))
```

```python
result=pd.DataFrame(data=predict,columns=['IsBadBuy'])
result.head()
```

```python
result['RefId']=test_df['RefId']
result.set_index('RefId',inplace=True)
```
