1 not encoding can be applied at categorial columns. So let's us first create a columns which are cotegorial 

```python
categorial = train_data.drop(['RefId','IsBadBuy'], axis=1).columns[train_data.drop(['RefId','IsBadBuy'], axis=1).dtypes=='object']
```

```python
categorial
```

```python
train_data[categorial[0]]    #3 categories
```

This below called 1 not encoding. 1 not encoding will convert the scalar values into the vector values.

```python
pd.get_dummies(train_data[categorial[0]])  # 3 COLUMNS    #this called 1 not encoding 
```

Changing column name or we can say rename

```python
for i in categorial:
    dummies = pd.get_dummies(train_data[i])
    dummies.columns=str(i)+'_'+dummies.columns
    train_data=pd.concat([train_data,dummies],axis=1)
    train_data.drop(i,inplace=True,axis=1)
    dummies=pd.get_dummies(test_data[i])
    dummies.columns=str(i)+'_'+dummies.columns
    test_data=pd.concat([test_data,dummies],axis=1)
    test_data.drop(i,inplace=True,axis=1)    
```

```python
from sklearn.model_selection import train_test_split
X=train_data.drop(['RefId','IsBadBuy'], axis=1)
y=train_data['IsBadBuy']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42)
```

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
KNN=KNeighborsClassifier(n_neighbors=11)
KNN.fit(X_train,y_train)
```

```python
KNN.score(X_test,y_test)
```

```python
predict=KNN.predict(test_data.drop('RefId',axis=1))
```

```python
Submission = pd.DataFrame(data=predict,columns=['IsBadBuy'])
Submission.head()
```

```python
Submission['RefId']=test_data['RefId']
Submission.set_index('RefId', inplace=True)
```
