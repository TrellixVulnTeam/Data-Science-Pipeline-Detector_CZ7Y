```python
train_df = pd.read_csv("../input/google-quest-challenge/train.csv")
train_df
```

```python
matrix = train_df.iloc[:, -30:].corr()
columns = matrix.columns
matrix = matrix.values
```

```python
# 重み付きグラフのファイル生成
with open("graph.txt", mode='w') as f:
    for i in range(30):
        for j in range(i+1):
            if i == j:
                continue
            v = matrix[i][j]
            if abs(v) < 0.4:
                continue
            f.write(f"{i} {j} {v:.2g}\n")
```

```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
G = nx.DiGraph()

# 重み付きのファイルの読み込み
G = nx.read_weighted_edgelist('graph.txt', nodetype=int)

# レイアウトと頂点の色を適当に設定
pos = nx.spring_layout(G, k=3)
node_color = ['r' if i < 2 else 'b' for i in range(G.number_of_edges())]

# グラフの描画
edge_labels = {(i, j): w['weight'] for i, j, w in G.edges(data=True)}
nx.draw_networkx_edge_labels(G,pos, edge_labels=edge_labels)
nx.draw_networkx(G, pos)

# 表示
plt.axis("off")
plt.show()
```

```python
# number of annotators
for column in train_df.iloc[:, -30:].columns:
    s = set(train_df[column].values)
    d = 10
    for v in s:
        for u in s:
            if v != u:
                d = min(abs(v - u), d)
    for i in range(100):
        if d * i >= 1:
            print(column, i)
            break
```
