```python
documents_categories = pd.read_csv("../input/documents_categories.csv")
unique_categories = pd.Series.unique(documents_categories['category_id'])
groupbydocid = documents_categories.groupby(['document_id'])
```

```python
category_confidence = groupbydocid.agg(pd.Series.tolist)[['category_id', 'confidence_level']]
```
