##What does Outbrain do?

When you are serving the web (say you are on CNN website), after you finish reading your news there will be rows of links at the bottom of the article, showing relevant stories. Those recommendations are delivered by Ourbrain. Outbrain promoted these stories from advertiser on publisher websites.
