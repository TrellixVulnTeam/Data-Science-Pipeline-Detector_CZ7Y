## Find Outliers

```python
def get_cat_idx():
    cat_column_index=[]
    for i in range(len(train.columns)):
        if(train.iloc[:,i].dtype=="O"):
            cat_column_index.append(i)
    return cat_column_index    

def get_num_idx():
    num_column_index=[]
    for i in range(len(train.columns)):
        if(train.iloc[:,i].dtype!="O"):
            num_column_index.append(i)
    return num_column_index 

train_cat_idx = get_cat_idx()  
train_num_idx = get_num_idx()
```

```python
def detect_outliers(df,features):
    outlier_indices = []
    
    for c in features:
        Q1 = np.percentile(df[c],25)
        Q3 = np.percentile(df[c],75)
        IQR = Q3 - Q1
        outlier_step = IQR * 1.5
        outlier_list_col = df[(df[c] < Q1 - outlier_step) | (df[c] > Q3 + outlier_step)].index
        outlier_indices.extend(outlier_list_col)
    
    outlier_indices = Counter(outlier_indices)
    multiple_outliers = list(i for i, v in outlier_indices.items() if v > 10)
    
    return multiple_outliers
```

```python
col_num_name=[]
for i in train_num_idx:
    colname = train.columns[i]
    col_num_name.append(colname)

col_num_name.remove('Id')
```

## Backward Elemination

```python
def backward_elimination(data, target,significance_level = 0.05):
    features = data.columns.tolist()
    while(len(features)>0):
        features_with_constant = sm.add_constant(data[features])
        p_values = sm.OLS(target, features_with_constant).fit().pvalues[1:]
        max_p_value = p_values.max()
        if(max_p_value >= significance_level):
            excluded_feature = p_values.idxmax()
            features.remove(excluded_feature)
        else:
            break 
    return features
```

```python
feat=backward_elimination(X_train,y_train)
```

```python
X_train = X_train.loc[:,feat]
X_test = test.loc[:,feat]
```
