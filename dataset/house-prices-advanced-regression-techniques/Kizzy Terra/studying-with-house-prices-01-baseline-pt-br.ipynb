# Introdução ao Aprendizado de Máquina

Nessa aula iremos treinar nosso modelo baseline, ignorando variáveis categóricas, sem seleção das malhores features e sem otimização de parâmetros. Essa implemenntação servirá como nosso primeiro exercício prático de modelagem e também nos oferecerá uma referência para as próximas aulas em que faremos melhorias na modelagem.

## Carregando os dados

```python
X_full = pd.read_csv('/kaggle/input/house-prices-advanced-regression-techniques/train.csv', index_col='Id')
X_full_test = pd.read_csv('/kaggle/input/house-prices-advanced-regression-techniques/test.csv', index_col='Id')
```

```python
X_full.head()
```

```python
X_full.info()
```

## Excluindo variáveis categóricas

```python
X_full = X_full.select_dtypes(exclude=['object'])
X_full_test = X_full_test.select_dtypes(exclude=['object'])
```

```python
X_full.info()
```

```python
X_full_test.info()
```

```python
X = X_full.drop(['SalePrice'], axis=1)
y = X_full.SalePrice
X_test = X_full_test
# y = f(X)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.impute import SimpleImputer
```

## Substituição de valores ausentes

```python
imp_mean = SimpleImputer(strategy='mean')
X= imp_mean.fit_transform(X)
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, train_size=0.8, random_state=3)
```

```python
X_test = imp_mean.transform(X_test)
```

## Treinando uma Árvore de Decisão

```python
house_prices_model = DecisionTreeRegressor(random_state=0)
```

```python
house_prices_model.fit(X_train,y_train)
```

```python
val_predictions = house_prices_model.predict(X_val)
```

## Avaliando o modelo

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_val, val_predictions, squared=False)
```

```python
X_full.describe().SalePrice
```

```python
test_predictions = house_prices_model.predict(X_test)
```

```python
test_predictions
```

## Criando arquivo de submissão

```python
output = pd.DataFrame({'Id': X_full_test.index,
                      'SalePrice': test_predictions})
output.set_index('Id', inplace=True)
output.to_csv('submission.csv')
```
