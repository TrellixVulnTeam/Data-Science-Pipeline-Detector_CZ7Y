<h1>Importing Libraries</h1>

```python
import pandas as pd
from sklearn.metrics import mean_squared_log_error
from lightautoml.automl.presets.tabular_presets import TabularAutoML
from lightautoml.tasks import Task
```

<h1>Reading the Data</h1>

```python
df_train, df_test = pd.read_csv('../input/house-prices-advanced-regression-techniques/train.csv'), pd.read_csv('../input/house-prices-advanced-regression-techniques/test.csv')
```

<h1> The Magic </h1>

```python
automl = TabularAutoML(task = Task('reg', metric = lambda y_true, y_pred: mean_squared_log_error(y_true, y_pred)))
oof_pred = automl.fit_predict(df_train,  roles = {'target': 'SalePrice', 'drop': ['Id']})
test_pred = automl.predict(df_test)
```

<h1>Writing the Output </h1>

```python
test_pred.data[:, 0]
```

```python
pd.DataFrame({'Id':df_test.Id, 'SalePrice': (test_pred.data[:, 0])}).to_csv('submit.csv', index = False)
```
