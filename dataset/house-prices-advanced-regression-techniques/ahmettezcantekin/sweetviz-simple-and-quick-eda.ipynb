# ****Introduction****

Exploratary Data Analysis (EDA) is one of the most significant part of the modelling which helps summarizing the main characteristics of the dataset often using visual methods. ,
At the beginning, if you aren't familiar with the data which you will work on it, it takes a lot of time for understanding the main characteristics of the data.

Nowadays, there are a lot of libraries which automate the process of EDA. Sweetviz is an automated EDA library which helps data scientists with beautiful and introductory visualizations.

Let's have a look at the Sweetviz!!!

```python
#Importing the necessary libraries!

import pandas as pd
import sweetviz as sv
from IPython.display import Image
```

```python
#Read the data
train=pd.read_csv('/kaggle/input/house-prices-advanced-regression-techniques/train.csv')
test=pd.read_csv('/kaggle/input/house-prices-advanced-regression-techniques/test.csv')
```

Sweetviz has 3 main functions which are analyze, compare and compare_intra. I will introduce you analyze and compare functions which are crucial features of Sweetviz!

Let's start with analyze function!

```python
#analyzing the dataset
df_analysis=sv.analyze(train)
```

```python
#show analysis as html
df_analysis.show_html('train_analysis.html')
```

```python
Image("../input/sweetviz/sweetviz.JPG")
```

Compare is another function of Sweetviz which compares 2 different datasets.

Let's have a look at the compare function.

```python
df_1=train.iloc[:,0:10].head(100)
df_2=train.iloc[:,0:10].tail(100)

df_compare = sv.compare(df_1, df_2)
df_compare.show_html('comparison_result.html')
```

```python
Image("../input/sweetvizcompare/sweetviz_compare.JPG")
```
