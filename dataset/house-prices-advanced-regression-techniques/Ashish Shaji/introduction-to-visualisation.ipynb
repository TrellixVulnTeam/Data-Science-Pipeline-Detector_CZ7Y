# Bar Plot

A bar chart or bar graph is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent. The bars can be plotted vertically or horizontally.


1. using Seaborn libary

```python
# using data from Titanic dataset
sns.barplot(x="Sex", y="Survived",data=df_titanic)
```

> 2. Using matplotlib Libary

```python
plt.subplots(figsize=(10,5))
data=df_titanic.copy()
data.groupby(df_titanic["Parch"])['Survived'].median().plot.bar()
plt.xlabel("Parch")
plt.ylabel('Survived')
plt.show()
```

```python
# using house price data
data=df_house_price.copy()
data.groupby(df_house_price["MSSubClass"])['SalePrice'].median().plot.bar()
plt.xlabel("MSSubClass")
plt.ylabel('SalePrice')
plt.show()
```

# Count Plot
Count Plot shows the number of occurrences of an item based on a certain type of category.

```python
ax = sns.countplot(x="Survived", data=df_titanic)
```

# Pie Chart
Pie charts are a visual way of displaying data that might otherwise be given in a small table. Pie charts are useful for displaying data that are classified into nominal or ordinal categories.

```python
plt.subplots(figsize=(5,5))
a=df_titanic.groupby(["Sex"]).Sex.count()
colors = ['gold', 'green']

a.plot.pie(colors=colors,autopct='%1.1f%%',shadow=True, startangle=175)
```

```python
a=df_titanic.groupby(["Pclass"]).Pclass.count()
colors = ['red', 'green',"yellow"]

a.plot.pie(colors=colors,autopct='%1.1f%%',shadow=True, startangle=175)
```

# Scatter Plot

# BoxPlot
For some distributions/datasets, you will find that you need more information than the measures of central tendency (median, mean, and mode). Boxplots are a standardized way of displaying the distribution of data based on a five number summary (“minimum”, first quartile (Q1), median, third quartile (Q3), and “maximum”).


```python

f,ax=plt.subplots(1,2,figsize=(20,7))

sns.boxplot(x=df_titanic["Age"],color="red",ax=ax[0]) 
sns.boxplot(x ='Pclass', y='Age',ax=ax[1],data=df_titanic) 
```

```python
f,ax=plt.subplots(figsize=(13,7))
sns.boxplot(x ='Pclass', y='Age',hue="Survived",data=df_titanic, palette="bone") 
```

# stripplot
A strip plot is a scatter plot where one of the variables is categorical. They can be combined with other plots to provide additional information.

```python
f,ax=plt.subplots(1,2,figsize=(20,7))

sns.stripplot(x ='Fare', y='Survived',ax=ax[0],data=df_titanic,orient="h",palette="rocket") 
ax[0].set_title('Survived V/s Fare Horizontal Strip plot')
ax[1].set_title('Age V/s Pclass Vertical Strip plot')
sns.stripplot(x ='Pclass', y='Age',ax=ax[1],data=df_titanic) 
```

# swarmplot
This function is similar to stripplot, but the points are adjusted (only along the categorical axis) so that they don't overlap. This gives a better representation of the distribution of values, 

```python
fig, ax = plt.subplots(figsize=(14,7)) 
sns.swarmplot(x ='Pclass', y='Age',ax=ax,size=4,data=df_titanic, palette="YlOrRd") 
```

Color the points using a second categorical variable "Sex"

```python
fig, ax = plt.subplots(figsize=(15,8))
sns.swarmplot(x ='Pclass', y='Age',ax=ax,hue="Sex",size=4,data=df_titanic) 
```

Draw swarms of observations on top of a box plot

```python
fig, ax = plt.subplots(figsize=(15,8))

sns.swarmplot(x ='Survived', y='Age',ax=ax,size=4,data=df_titanic, palette="PuRd") 
sns.boxplot(x ='Survived', y='Age',ax=ax,data=df_titanic) 
```

# Boxen Plot
The Boxen plot is very similar to box plot, except for the fact that it plots different quartile values. By plotting different quartile values, we are able to understand the shape of the distribution particularly in the head end and tail end

```python
fig, ax = plt.subplots(figsize=(15,8))
sns.boxenplot(x ='Pclass', y='Age',ax=ax,hue="Sex",data=df_titanic,palette="magma_r") 

```

# Violin plots
A violin plot is a method of plotting numeric data. It is similar to a box plot, with the addition of a rotated kernel density plot on each side. 

```python
fig, ax = plt.subplots(2,1,figsize=(15,8))
sns.violinplot(x=df_titanic["Age"],ax=ax[0],palette="winter")
ax[0].set_title('Plotting Age ')
ax[0].set_title('Plotting Pclass')
sns.violinplot(x=df_titanic["Pclass"],ax=ax[1],palette="tab20c_r")
```

```python
fig, ax = plt.subplots(2,1,figsize=(15,10))

ax[0].set_title('Pclass v/s Age Voilen plot')
sns.violinplot(x="Pclass",y="Age",ax=ax[0],palette="tab20c_r",data=df_titanic)

ax[1].set_title('Pclass v/s Age Voilen plot differeniting by gender')
sns.violinplot(x ='Pclass', y='Age',data=df_titanic,ax=ax[1], hue="Sex",palette="Set2") 
```

# Point Plot
Point plots can be more useful than bar plots for focusing comparisons between different levels of one or more categorical variables.A point plot represents an estimate of central tendency for a numeric variable by the position of scatter plot points and provides some indication of the uncertainty around that estimate using error bars.

```python
fig, ax = plt.subplots(figsize=(13,6))
sns.pointplot(x="Survived",y="Age",hue="Sex",ax=ax,palette="winter",data=df_titanic)
```

# Correlation Heatmap
Correlation is a statistical technique that can show whether and how strongly pairs of variables are related will help us to find the most coorelated feature.we are using corelation heat map and Zoomed heatmap



```python
corrmat = df_house_price.corr()
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corrmat, vmax=.8,square=True)
```

# Zoomed Heatmap

```python
k= 11
cols = corrmat.nlargest(k,'SalePrice')['SalePrice'].index

cm = np.corrcoef(df_house_price[cols].values.T)
f , ax = plt.subplots(figsize = (14,12))
sns.heatmap(cm, vmax=.8, linewidths=0.01,square=True,annot=True,cmap='viridis',
            linecolor="white",xticklabels = cols.values ,annot_kws = {'size':12},yticklabels = cols.values)
```

The missingno correlation heatmap measures nullity correlation: how strongly the presence or absence of one variable affects the presence of another:

```python
import missingno as msno
msno.heatmap(df_house_price,cmap='flag')
```
