```python
y_train=trdt.iloc[:,1]
X_train=trdt.iloc[:,2:].drop(['Name','Cabin','Ticket'],axis=1)
print("X_train:")
print(X_train.head())
print("y_train:")
print(y_train.head())
```

```python
X_train=pd.get_dummies(X_train,drop_first=True)
X_train
```

```python
from sklearn.impute import SimpleImputer
Imp=SimpleImputer(strategy='median',missing_values=np.nan)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
```

```python
from sklearn.pipeline import Pipeline
steps=[('imp',Imp),('rf',rfc)]
pipe=Pipeline(steps)
```

```python
from sklearn.model_selection import GridSearchCV
parameter_space={'rf__max_depth':np.arange(3,10,1),'rf__n_estimators':np.arange(50,150,10)}
cvmod=GridSearchCV(pipe ,parameter_space, cv=5)
```

```python
cvmod.fit(X_train,y_train)
```

```python
cvmod.best_score_
```

```python
cvmod.best_params_
```
