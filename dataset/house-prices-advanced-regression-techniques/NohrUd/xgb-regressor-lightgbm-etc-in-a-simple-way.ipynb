## 10. LightGBM Model

#### 1)Preparation for LightGBM

```python
#make train data
x_train_p=df.iloc[0:1160,1:80]
y_train_p=df.iloc[0:1160,-1]
x_train=x_train_p.to_numpy()
y_train=y_train_p.to_numpy()
#make train evaluation data
x_train_eval_p=df.iloc[1160:,1:80]
y_train_eval_p=df.iloc[1160:,-1]
x_train_eval=x_train_p.to_numpy()
y_train_eval=y_train_p.to_numpy()
#make test data
x_test_p=test_data.iloc[0:,1:]
x_test=x_test_p.to_numpy()
```

### 2)Let's do it. 

```python
#Model3 lightgbm
lgb_train = lgb.Dataset(x_train, y_train)
lgb_eval = lgb.Dataset(x_train_eval, y_train_eval, reference=lgb_train) 
#parameters
params = {'task': 'train','boosting_type': 'gbdt','objective': 'regression','metric': {'rmse'}}
#Start training
model3 = lgb.train(params,train_set=lgb_train,valid_sets=lgb_eval)
#Check the model performance
y_pred=model3.predict(x_train_eval)
r2 = r2_score(y_train_eval,y_pred)
print("The score on traing data of this machine learning model3 is ",r2,"!")
```

### <i>Which is better XGB boost or LightGBM? LightGBM?!</i>

```python
#test_predction
y_pred=model3.predict(x_test)
```
