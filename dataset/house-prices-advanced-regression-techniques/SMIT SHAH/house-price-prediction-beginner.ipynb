# Dealing With Categorical Data

* Creating Dummies for Train Data

```python
LandContour = pd.get_dummies(train_data['LandContour'],drop_first=True)
MSZoning = pd.get_dummies(train_data['MSZoning'],drop_first=True)
LotShape = pd.get_dummies(train_data['LotShape'],drop_first=True)
Neighborhood = pd.get_dummies(train_data['Neighborhood'],drop_first=True)
HouseStyle = pd.get_dummies(train_data['HouseStyle'],drop_first=True)
BldgType = pd.get_dummies(train_data['BldgType'],drop_first=True)
Foundation = pd.get_dummies(train_data['Foundation'],drop_first=True)
CentralAir = pd.get_dummies(train_data['CentralAir'],drop_first=True)
RoofStyle = pd.get_dummies(train_data['RoofStyle'],drop_first=True)
KitchenQual = pd.get_dummies(train_data['KitchenQual'],drop_first=True)
Functional = pd.get_dummies(train_data['Functional'],drop_first=True)
GarageType = pd.get_dummies(train_data['GarageType'],drop_first=True)
SaleType = pd.get_dummies(train_data['SaleType'],drop_first=True)
SaleCondition = pd.get_dummies(train_data['SaleCondition'],drop_first=True)
LotConfig = pd.get_dummies(train_data["LotConfig"], drop_first=True)
BsmtFinType1 = pd.get_dummies(train_data["BsmtFinType1"], drop_first=True)
```

```python
train_data.drop(['LandContour','MSZoning','LotShape', 
                 'Neighborhood', 'HouseStyle', 'BldgType',
                 'Foundation', 'CentralAir', 'RoofStyle', 
                 'KitchenQual', 'Functional', 'GarageType', 
                 'SaleType', 'SaleCondition', 'LotConfig','LandSlope', 'BsmtFinType1'], axis = 1, inplace=True)
```

```python
train_data = pd.concat([train_data,LandContour,MSZoning,LotShape,Neighborhood,HouseStyle,BldgType,Foundation,CentralAir,RoofStyle,KitchenQual,Functional,GarageType,SaleType,SaleCondition, LotConfig, BsmtFinType1],axis=1)
```

* Creating Dummies for test data

```python
LandContour = pd.get_dummies(test_data['LandContour'],drop_first=True)
MSZoning = pd.get_dummies(test_data['MSZoning'],drop_first=True)
LotShape = pd.get_dummies(test_data['LotShape'],drop_first=True)
Neighborhood = pd.get_dummies(test_data['Neighborhood'],drop_first=True)
HouseStyle = pd.get_dummies(test_data['HouseStyle'],drop_first=True)
BldgType = pd.get_dummies(test_data['BldgType'],drop_first=True)
Foundation = pd.get_dummies(test_data['Foundation'],drop_first=True)
CentralAir = pd.get_dummies(test_data['CentralAir'],drop_first=True)
RoofStyle = pd.get_dummies(test_data['RoofStyle'],drop_first=True)
KitchenQual = pd.get_dummies(test_data['KitchenQual'],drop_first=True)
Functional = pd.get_dummies(test_data['Functional'],drop_first=True)
GarageType = pd.get_dummies(test_data['GarageType'],drop_first=True)
SaleType = pd.get_dummies(test_data['SaleType'],drop_first=True)
SaleCondition = pd.get_dummies(test_data['SaleCondition'],drop_first=True)
LotConfig = pd.get_dummies(test_data["LotConfig"], drop_first=True)
BsmtFinType1 = pd.get_dummies(test_data["BsmtFinType1"], drop_first=True)
```

```python
test_data.drop(['LandContour','MSZoning','LotShape', 
                 'Neighborhood', 'HouseStyle', 'BldgType',
                 'Foundation', 'CentralAir', 'RoofStyle', 
                 'KitchenQual', 'Functional', 'GarageType', 
                 'SaleType', 'SaleCondition','LotConfig','LandSlope', 'BsmtFinType1'], axis = 1, inplace=True)
```

```python
test_data = pd.concat([test_data,LandContour,MSZoning,LotShape,Neighborhood,HouseStyle,BldgType,Foundation,CentralAir,RoofStyle,KitchenQual,Functional,GarageType,SaleType,SaleCondition, LotConfig, BsmtFinType1],axis=1)
```
