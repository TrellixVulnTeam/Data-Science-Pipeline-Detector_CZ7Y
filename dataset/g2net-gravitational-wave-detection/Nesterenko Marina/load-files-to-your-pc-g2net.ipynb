```python
import kaggle
import pandas as pd
import numpy as np

from kaggle.api.kaggle_api_extended import KaggleApi
# =============================================================================
# functions
# =============================================================================
def get_paths(dataset, i, is_train=True):
    """It creates a path to file on kaggle and path where to download"""
    _id = dataset.iloc[i]['id']
    target = dataset.iloc[i]['target']
    folder = 'train' if is_train else 'test'
    folder_0 = f'{_id[0]}'
    folder_1 = f'{_id[1]}'
    folder_2 = f'{_id[2]}'
    path_folder = f"{folder}/{folder_0}/{folder_1}/{folder_2}/" #path where to download
    path_from = f"{folder}/{folder_0}/{folder_1}/{folder_2}/{_id}.npy"
    api.competition_download_file(competition, path_from , path=path_folder, force = True)
    x = np.load(path_from)
    return x, target
# =============================================================================
# download files from kaggle to your PC
# =============================================================================
competition = 'g2net-gravitational-wave-detection'
path = 'dataset_main/' # your folder on PC

api = KaggleApi()
api.authenticate()

training_labels = pd.read_csv(path + 'training_labels.csv') # You need to have this file's already downloaded

range_from = 0
range_to = 100 # You can choose how much files to download

for i in range(range_from, range_to+1):
    x, target = get_paths(training_labels, i, is_train=True)


#print(x, target)
#print('len(x) = ', len(x))
#print('len(x[0]) = ', len(x[0]))
```
