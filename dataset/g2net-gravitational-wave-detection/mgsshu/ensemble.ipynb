```python
df_efnet_submit = pd.read_csv("../input/efficientnet-cqt-pytorch-lightning-starter/submission.csv")
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
df_efnet_submit["prob"] = df_efnet_submit["target"].apply(sigmoid)
```

```python
df_efnet_submit
```

```python
plt.figure(figsize=(12, 12))
sns.scatterplot(
    x=df_1dcnn_submit['target'], 
    y=df_efnet_submit['prob']
    )
plt.show()
```

```python
df_ike_1 = pd.read_csv("../input/ikesubmission/submission_v13_mod3.csv")
df_ike_2 = pd.read_csv("../input/ike-submission/submission_v14.csv")
```

```python
df_ensemble = df_efnet_submit[["id", "prob", "model2", "model3"]].copy()
df_ensemble.columns = ["id", "model1", "model2", "model3"]
df_ensemble['ike1'] = df_ike_1["target"].copy()
# df_ensemble['ike2'] = df_ike_2["target"].copy()
```

```python
df_ensemble
```

```python
df_ike_2
```

```python
df_ensemble_2 = pd.merge(df_ensemble, df_ike_2, how="inner", on="id")
```

```python
df_ensemble_2
```

```python
# simple mean

df_ensemble_2["simple_mean"] = df_ensemble_2[["model1", "model2", "model3", "ike1", "target"]].mean(axis=1).copy()
df_ensemble_2
```

```python
submission_simple_mean = df_ensemble_2[["id", "simple_mean"]]
submission_simple_mean.columns = ["id", "target"]
```

```python
submission_simple_mean.to_csv('submission_simple_mean.csv', index = False)
```
