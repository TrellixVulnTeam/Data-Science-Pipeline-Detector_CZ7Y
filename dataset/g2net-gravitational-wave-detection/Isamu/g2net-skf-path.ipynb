```python
gcs_paths = []
GCS_path = KaggleDatasets().get_gcs_path(f"g2net-skf")
gcs_paths.append(GCS_path)
print(GCS_path)
```

```python
all_files = []
for path in gcs_paths:
    all_files.extend(np.sort(np.array(tf.io.gfile.glob(path + "/train*.tfrecords"))))

print("train_files: ", len(all_files))
```
