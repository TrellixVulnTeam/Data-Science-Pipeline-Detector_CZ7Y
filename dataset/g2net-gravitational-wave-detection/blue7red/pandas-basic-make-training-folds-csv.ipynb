```python
import numpy as np
import pandas as pd 
from sklearn.model_selection import StratifiedKFold
```

# load training data 

```python
training_labels = pd.read_csv("../input/g2net-gravitational-wave-detection/training_labels.csv")
training_labels.head()
```

# get unique label count 

```python
training_labels.target.nunique()
```

# split folds for 5 fold training model

```python
skf = StratifiedKFold(n_splits=5,random_state=42, shuffle=True)
skfSplit = skf.split(training_labels, training_labels["target"])
fold = 1
for train_index, test_index in skfSplit:
    training_labels.loc[test_index,"fold"] = fold
    fold = fold + 1
    print("TRAIN:", train_index, "TEST:", test_index)
    
training_labels = training_labels.astype({'fold': 'int64'})

print(training_labels.info())
training_labels.to_csv("g2net_5folds.csv",index = False)
    
training_labels.head(20)
```

```python
result = pd.read_csv("./g2net_5folds.csv")
result.head()
```

# split folds for 10 fold training model

```python
skf = StratifiedKFold(n_splits=10,random_state=42, shuffle=True)
skfSplit = skf.split(training_labels, training_labels["target"])
fold = 1
for train_index, test_index in skfSplit:
    training_labels.loc[test_index,"fold"] = fold
    fold = fold + 1
    print("TRAIN:", train_index, "TEST:", test_index)
    
training_labels = training_labels.astype({'fold': 'int64'})

print(training_labels.info())
training_labels.to_csv("g2net_10folds.csv",index = False)
    
training_labels.head(20)
```
