```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
import os
```

```python
path = '/kaggle/input/g2net-gravitational-wave-detection/'
os.listdir(path)
```

```python
def convert_image_id_2_path(image_id: str, is_train: bool = True) -> str:
    folder = "train" if is_train else "test"
    return "../input/g2net-gravitational-wave-detection/{}/{}/{}/{}/{}.npy".format(
        folder, image_id[0], image_id[1], image_id[2], image_id 
    )
```

```python
train_labels = pd.read_csv(path+'training_labels.csv')
samp_subm = pd.read_csv(path+'sample_submission.csv')
```

```python
print('Number train samples:', len(train_labels))
print('Number submission samples:', len(samp_subm))
```

```python
def plot_data(data):
    """ Plot 3 Detections of data array"""
    
    fig, axs = plt.subplots(1, 3, figsize=(20, 5))
    axs = axs.ravel()
    for i in range(3):
        x = range(len(data[i]))
        y = data[i]
        axs[i].plot(x, y)
        axs[i].grid()
        axs[i].set_title('Detection '+str((i+1)))
```

```python
id_ = train_labels.loc[0, 'id']
id_
path_in = convert_image_id_2_path(id_, True)
```

```python
data_array = np.load(path_in)
data_array.shape
print(data_array)
```

```python
plot_data(data_array)
```

```python
sns.countplot(data=train_labels, x="target")
```

```python
def visualize_sample(
    _id, 
    target, 
    colors=("black", "red", "green"), 
    signal_names=("LIGO Hanford", "LIGO Livingston", "Virgo")
):
    path = convert_image_id_2_path(_id)
    x = np.load(path)
    plt.figure(figsize=(16, 7))
    for i in range(3):
        plt.subplot(4, 1, i + 1)
        plt.plot(x[i], color=colors[i])
#         plt.title(signal_name[i], fontsize=14)
#         plt.xlabel("time")
        plt.legend([signal_names[i]], fontsize=12, loc="lower right")
        
        plt.subplot(4, 1, 4)
        plt.plot(x[i], color=colors[i])
    
    plt.subplot(4, 1, 4)
    plt.legend(signal_names, fontsize=12, loc="lower right")

    plt.suptitle(f"id: {_id} target: {target}", fontsize=16)
    plt.show()
```

```python
for i in random.sample(train_labels.index.tolist(), 3):
    _id = train_labels.iloc[i]["id"]
    target = train_labels.iloc[i]["target"]

    visualize_sample(_id, target)
```
