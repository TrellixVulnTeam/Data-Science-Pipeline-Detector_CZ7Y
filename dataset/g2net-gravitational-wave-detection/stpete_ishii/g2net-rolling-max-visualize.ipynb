```python
import numpy as np
import pandas as pd
import os
from PIL import Image
import matplotlib.pyplot as plt
from tqdm import tqdm
```

# Data preparation

```python
trainpaths=[]
for dirname, _, filenames in os.walk('../input/g2net-gravitational-wave-detection/train'):
    for filename in filenames:
        trainpaths+=[(os.path.join(dirname, filename))]
```

```python
print(len(trainpaths))
sorted(trainpaths)[0:5]
```

# Data visualize

```python
# fig of original data
aa=np.load(trainpaths[2])
figure,axis = plt.subplots(1,3,figsize=(18,5),)
axis[0].set_title("original0")
axis[0].set_xlabel(aa[0].shape)
axis[0].set_ylabel(aa[0].size)
axis[0].plot(aa[0],c="g")
axis[1].set_title("original1")
axis[1].set_xlabel(aa[1].shape)
axis[1].set_ylabel(aa[1].size)
axis[1].plot(aa[1],c="g")
axis[2].set_title("original2")
axis[2].set_xlabel(aa[2].shape)
axis[2].set_ylabel(aa[2].size)
axis[2].plot(aa[2],c="g")
```

```python
# fig of lighteded data
aa1=np.zeros((3,128))
for i in range(0,32*128,32):
    aa1[:,i//32]+=aa[:,i]
figure,axis = plt.subplots(1,3,figsize=(18,5),)
axis[0].set_title("lightened 0")
axis[0].set_xlabel(aa1[0].shape)
axis[0].set_ylabel(aa1[0].size)
axis[0].plot(aa1[0],c="g")
axis[1].set_title("lightend 1")
axis[1].set_xlabel(aa1[1].shape)
axis[1].set_ylabel(aa1[1].size)
axis[1].plot(aa1[1],c="g")
axis[2].set_title("lightend 2")
axis[2].set_xlabel(aa1[2].shape)
axis[2].set_ylabel(aa1[2].size)
axis[2].plot(aa1[2],c="g")
```

```python
aa1t=pd.DataFrame(aa1).T
aa1t
```

```python
aa1t=pd.DataFrame(aa1).T
aa1t['0 max']=aa1t.loc[:,0].rolling(window=8).max()
aa1t['1 max']=aa1t.loc[:,1].rolling(window=8).max()
aa1t['2 max']=aa1t.loc[:,2].rolling(window=8).max()
aa1t
```

```python
# fig of rolling max data
aa2=np.array(aa1t.T)
figure,axis = plt.subplots(1,3,figsize=(18,5),)
axis[0].set_title("rolling max 0")
axis[0].set_xlabel(aa2[0].shape)
axis[0].set_ylabel(aa2[0].size)
axis[0].plot(aa2[3],c="g")
axis[1].set_title("rolling max 1")
axis[1].set_xlabel(aa2[0].shape)
axis[1].set_ylabel(aa2[0].size)
axis[1].plot(aa2[4],c="g")
axis[2].set_title("rolling max 2")
axis[2].set_xlabel(aa2[0].shape)
axis[2].set_ylabel(aa2[0].size)
axis[2].plot(aa2[5],c="g")
```
