## Easy way to plot the different gravitational waves

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from PIL import Image
import matplotlib.pyplot as plt
```

```python
def plot_series(path):
    labels = pd.read_csv("../input/g2net-gravitational-wave-detection/training_labels.csv")
    data = np.load(path)
    name = (path.split("/")[-1]).split(".")[0]
    color = ['red','blue','green']
    signal_names = ["LIGO Hanford", "LIGO Livingston", "Virgo"]
    fig, axs = plt.subplots(3,1,figsize=(13,11))
    plt.suptitle("Target : {}".format(labels[labels["id"] == name].iloc[0][1]), size=16)
    for i in range(3):
        axs[i].plot(data[i], color = color[i])
        axs[i].title.set_text(signal_names[i])
```

```python
plot_series("../input/g2net-gravitational-wave-detection/train/4/3/0/4302103bb1.npy")
```
