#### Adding Imports

```python
import pandas as pd
from sklearn import model_selection
```

#### Read the data

```python
data = pd.read_csv("../input/g2net-gravitational-wave-detection/training_labels.csv")
data.head()
```

Set a new column in the data frame to split the data onto k folds


```python
data["k_fold_splitting"] = -1
```

n_splits : No of splits we want to make of the data

shuffle : Whether to shuffle each classâ€™s samples before splitting into batches.

random_state: When shuffle is True, random_state affects the ordering of the indices, which controls the randomness of each fold for each class. 

- **Pass an int for reproducible output across multiple function calls**

```python
k_fold = model_selection.StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for fold, (_, value) in enumerate(k_fold.split(X=data, y=data.target.values)):
    data.loc[value, 'k_fold_splitting'] = fold
```

A new column **k_fold_splitting** is created which contains the fold of each row.

```python
data.head()
```

Converting the data frame to a csv file.

```python
data.to_csv("k_fold_train_folds.csv", index=False)
```
