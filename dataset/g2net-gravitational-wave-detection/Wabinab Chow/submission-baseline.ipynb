```python
from fastai.vision.all import *

def predict_batch(self, item, rm_type_tfms=None, with_input=False, num_workers=0):
    dl = self.dls.test_dl(item, rm_type_tfms=rm_type_tfms, num_workers=num_workers)
    ret = self.get_preds(dl=dl)
    return ret

Learner.predict_batch = predict_batch
```

```python
# Efficientnet B4 (effb7_ft5.pkl)  ==> INCORRECT NAMING USED DUE TO CONFUSION...

class EffNetModel(nn.Module):
    def __init__(self):
        super(EffNetModel, self).__init__()
        self.net = EfficientNet.from_pretrained("efficientnet-b4")
        n_features = self.net._fc.in_features
        self.net._fc = nn.Linear(in_features=n_features, out_features=2, bias=True)
        
    def forward(self, x):
        out = self.net(x)
        return out


cnn_model = EffNetModel()
```
