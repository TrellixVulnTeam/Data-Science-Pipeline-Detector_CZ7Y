## Visulaizing the Q Transform

In mathematics and signal processing, the constant-Q transform, simply known as CQT transforms a data series to the frequency domain. It is related to the Fourier transform and very closely related to the complex Morlet wavelet transform

As the black holes spiral closer and closer in together, the frequency of the gravitational waves increases. Scientists call these sounds "chirps," because some events that generate gravitation waves would sound like a bird's chirp

[Gravitational wave sound from caltech's website](http://https://www.ligo.caltech.edu/video/ligo20160211v2)

## What is the difference between FFT and Constant Q transfrom ? 

A standard DFT uses a constant buffer size throughout all frequencies. This typically leads to a pretty consistent, fully continuous transform. However, the constant bin size for all frequencies leads to some problems when you map frequency on a logarithmic scale. Specifically, peaks on the lower end are incredibly wide (sometimes up to half an octave), lacking any sort of detail.

This is an issue for emulating human perception because humans perceive frequency on a logarithmic scale.

A Constant Q transform seeks to solve this problem by increasing the buffer size for lower frequencies, and alleviate some of the computational strain caused by this by reducing the buffer size used for high frequencies. It's pretty effective at this, but has a few drawbacks.[answer is here ](http://https://dsp.stackexchange.com/questions/43811/what-is-the-difference-between-constant-q-transform-and-wavelet-transform-and-wh)

```python
import librosa
import librosa.display
```

```python
def visualize_sample_mfcc(
    _id, 
    target,
    signal_names=("LIGO Hanford", "LIGO Livingston", "Virgo"),
    sr=2048,
):
    x = np.load(_id)
    plt.figure(figsize=(16, 5))
    for i in range(3):
        mfccs = librosa.feature.mfcc(x[i] / x[i].max(), sr=sr)
        print(max(mfccs[1]))
        plt.subplot(1, 3, i + 1)
        librosa.display.specshow(mfccs, sr=sr, x_axis="time", vmin=-200, vmax=60, cmap="coolwarm")
        plt.title(signal_names[i], fontsize=14)
        plt.colorbar()

 

    plt.show()
```

```python
from sklearn.decomposition import PCA
def mfcc_transformation(_id,sr=2048):
    mfcc_coef=[]
    x = np.load(_id)
    for i in range(3):
        mfccs = librosa.feature.mfcc(x[i] / x[i].max(), sr=sr)
        pca = PCA(n_components=3)
        principalComponents = pca.fit_transform(mfccs)
        mfcc_coef.append(principalComponents)
    return mfcc_coef
```

```python
NUM_BATCHES = 1
BATCH_SIZE = train_df.shape[0]/NUM_BATCHES
```

```python
def batch_transform(data_df, is_train=True):
    waves_data = []
    for path in tqdm(data_df.path, total=data_df.shape[0]):
        wave=np.load(path) 
        waves_data.append(wave)
    X = np.array(waves_data)
    nsamples, nx, ny = X.shape
    d2_X=X.reshape((nsamples,nx*ny))
    if is_train:        
        y= data_df['target'].values
        #     y = to_categorical(y) ## If using a softmax layer for prediction
        return d2_X,y
    else:
        return d2_X
```
