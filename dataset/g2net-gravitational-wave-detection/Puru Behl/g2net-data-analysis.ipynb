```python
# Making another  dataset now this time it's gonna be having the data for the different files
def get_df_numpy(path) :
    df=pd.DataFrame()
    data=np.load(path)
    df['site1']=data[0]
    df['site2']=data[1]
    df['site3']=data[2]
    
    return df
```

```python
def make_plot(data=data,id_1=0,id_0=1,types="histogram") :
    target_1=get_df_numpy(data['paths'][id_1])
    target_0=get_df_numpy(data['paths'][id_0])
    if types=="histogram" :
        fig = make_subplots(rows=1, cols=2)

        fig.add_trace(go.Histogram(x=target_1['site1'],name="Site 1"),row=1,col=1)
        fig.add_trace(go.Histogram(x=target_1['site2'],name= "Site 2"),row=1,col=1)
        fig.add_trace(go.Histogram(x=target_1['site3'],name="Site 3"),row=1,col=1)

        fig.add_trace(go.Histogram(x=target_0['site1'],name="Site 1"),row=1,col=2)
        fig.add_trace(go.Histogram(x=target_0['site2'],name= "Site 2"),row=1,col=2)
        fig.add_trace(go.Histogram(x=target_0['site3'],name="Site 3"),row=1,col=2)

        fig.update_layout(height=800, width=1200,barmode="stack",template="presentation")
        return fig
    if types=="line" :
        fig = make_subplots(rows=2, cols=1)

        fig.add_trace(go.Line(y=target_1['site1'],name="Site 1",x=[i for i in range(4096)]),row=1,col=1)
        fig.add_trace(go.Line(y=target_1['site2'],name= "Site 2",x=[i for i in range(4096)]),row=1,col=1)
        fig.add_trace(go.Line(y=target_1['site3'],name="Site 3",x=[i for i in range(4096)]),row=1,col=1)

        fig.add_trace(go.Line(y=target_0['site1'],name="Site 1",x=[i for i in range(4096)]),row=2,col=1)
        fig.add_trace(go.Line(y=target_0['site2'],name= "Site 2",x=[i for i in range(4096)]),row=2,col=1)
        fig.add_trace(go.Line(y=target_0['site3'],name="Site 3",x=[i for i in range(4096)]),row=2,col=1)

        fig.update_layout(height=800, width=1200,barmode="stack",template="presentation")
        return fig
    
    
def make_plot_from_array(average_1,average_0,types="histogram") :
    if types=="histogram" :

        fig = make_subplots(rows=1, cols=2)

        fig.add_trace(go.Histogram(x=average_1[0],name="Site 1"),row=1,col=1)
        fig.add_trace(go.Histogram(x=average_1[1],name= "Site 2"),row=1,col=1)
        fig.add_trace(go.Histogram(x=average_1[2],name="Site 3"),row=1,col=1)

        fig.add_trace(go.Histogram(x=average_0[0],name="Site 1"),row=1,col=2)
        fig.add_trace(go.Histogram(x=average_0[1],name= "Site 2"),row=1,col=2)
        fig.add_trace(go.Histogram(x=average_0[2],name="Site 3"),row=1,col=2)

        fig.update_layout(height=800, width=1200,barmode="stack",template="presentation")
        return fig
    if types=="line" :
        fig = make_subplots(rows=2, cols=1)

        fig.add_trace(go.Line(y=average_1[0],name="Site 1",x=[i for i in range(4096)]),row=1,col=1)
        fig.add_trace(go.Line(y=average_1[1],name= "Site 2",x=[i for i in range(4096)]),row=1,col=1)
        fig.add_trace(go.Line(y=average_1[2],name="Site 3",x=[i for i in range(4096)]),row=1,col=1)

        fig.add_trace(go.Line(y=average_0[0],name="Site 1",x=[i for i in range(4096)]),row=2,col=1)
        fig.add_trace(go.Line(y=average_0[1],name= "Site 2",x=[i for i in range(4096)]),row=2,col=1)
        fig.add_trace(go.Line(y=average_0[2],name="Site 3",x=[i for i in range(4096)]),row=2,col=1)

        fig.update_layout(height=800, width=1200,barmode="stack",template="presentation")
        return fig
```
