```python
import numpy as np
import pandas as pd

from sklearn import model_selection
```

```python
data = pd.read_csv("../input/g2net-gravitational-wave-detection/training_labels.csv")
```

```python
data.head()
```

```python
data["kfold"] = -1

kf = model_selection.StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for f, (t_, v_) in enumerate(kf.split(X=data, y=data.target.values)):
    data.loc[v_, 'kfold'] = f
```

```python
data.to_csv("train_folds.csv", index=False)
```
