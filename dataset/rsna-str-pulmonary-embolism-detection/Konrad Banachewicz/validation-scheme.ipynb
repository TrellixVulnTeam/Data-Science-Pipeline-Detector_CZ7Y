```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import LabelEncoder
```

```python
# settings

nfolds = 10

data_folder = '../input/rsna-str-pulmonary-embolism-detection/'
```

```python
# data
train = pd.read_csv(data_folder + 'train.csv')
test = pd.read_csv(data_folder + 'test.csv')

xfolds = train[['StudyInstanceUID', 'SeriesInstanceUID']].copy()
```

```python
# non-overlapping folds
le = LabelEncoder()
x = le.fit_transform(xfolds['StudyInstanceUID'])
xfolds['fold_st'] = x%nfolds

x = le.fit_transform(xfolds['SeriesInstanceUID'])
xfolds['fold_sr'] = x%nfolds
```

```python
xfolds.to_csv('folds.csv', index = False)
```
