### Import Libraries

```python
import pandas as pd
print('Loaded Imports')
```

### Load Data

```python
PATH = "../input/rsna-str-pulmonary-embolism-detection/"
train = pd.read_csv(PATH + "train.csv")
sub = pd.read_csv(PATH + "sample_submission.csv")
print('Loaded training and submission csv files.')
```

```python
feats = list(train.columns[3:5])+list(train.columns[8:12])+list(train.columns[13:17])
print('The list of relevant fields is', feats)
```

```python
train[feats].describe()
```

### Fill submission csv with mean values

```python
means = train[feats].mean().to_dict()
means
```

```python
sub['label'] = 0.2799
for feat in feats:
    sub.loc[sub.id.str.contains(feat, regex=False), 'label'] = means[feat]
sub.head()
```

### Submission

```python
sub.to_csv('submission.csv', index = False)
print('Saved submission.csv')
```
