```python
def compute_signals_per_stock(stk_id, signal_lst, mode='train', add_tif=True):
    book = pd.read_parquet(f'/kaggle/input/optiver-realized-volatility-prediction/book_{mode}.parquet/stock_id={stk_id}')
    trade = pd.read_parquet(f'/kaggle/input/optiver-realized-volatility-prediction/trade_{mode}.parquet/stock_id={stk_id}')
    trade = merge_asof(trade, book).set_index('time_id')
    book = book.set_index('time_id')
    
    bid_size1, ask_size1, bid_size2, ask_size2 = (book[col]/100.0 for col in ["bid_size1", "ask_size1", "bid_size2", "ask_size2"])
    bid_price1, ask_price1, bid_price2, ask_price2 = (book[col] for col in ["bid_price1", "ask_price1", "bid_price2", "ask_price2"])
    prev_book = pd.concat([bid_price1, ask_price1, bid_size1, ask_size1, bid_price2, ask_price2, bid_size2, ask_size2],axis=1).groupby('time_id').shift(1)
    prev_bid_price1, prev_ask_price1, prev_bid_size1, prev_ask_size1, prev_bid_price2, prev_ask_price2, prev_bid_size2, prev_ask_size2 = [prev_book[c] for c in prev_book.columns]
    book['mid'] = (ask_price1.values + bid_price1.values)/2.0
    book['spread_bps'] = ((ask_price1.values - bid_price1.values)/book['mid'].values)
    book['deep_spread_bps'] = ((ask_price2.values - bid_price2.values)/book['mid'].values)
    book['bid_spread_bps'] = ((bid_price1.values - bid_price2.values)/book['mid'].values)
    book['ask_spread_bps'] = ((ask_price2.values - ask_price1.values)/book['mid'].values)
    #book['tick_size'] = np.nanmax([(bid_price1- prev_bid_price1).abs().where(lambda x:x>0).values,(ask_price1 - prev_ask_price1).abs().where(lambda x:x>0).values], axis=)
    # tick_size = min((bid_price1 - prev_bid_price1).abs().where(lambda x:x>0).min(), (ask_price1 - prev_ask_price1).abs().where(lambda x:x>0).min())
    #print(tick_size)
    #book['raw_mid'] = book['mid'] / tick_size * 0.01
    #raw_ask_price1 = ask_price1.values / tick_size * 0.01
    #raw_bid_price1 = bid_price1.values / tick_size * 0.01
    #book['raw_spread_cents'] = raw_ask_price1 - raw_bid_price1
    #print(book[['raw_mid']].head(20))
    book['bid_ask_imb'] = (bid_size1.values - ask_size1.values)
    book['deep_imb'] = (bid_size1.values + bid_size2.values - ask_size1.values - ask_size2.values)
    book['time_diff'] = book['seconds_in_bucket'].groupby('time_id').apply(lambda x:x.diff().shift(-1)) # seconds in bucket may not start from 0 https://www.kaggle.com/c/optiver-realized-volatility-prediction/discussion/251775
    book['time_diff'] = book['time_diff'].fillna(600-book['seconds_in_bucket'])
    lvl1_size = bid_size1.values + ask_size1.values
    lvl2_size = bid_size2.values + ask_size2.values
    tot_size = lvl1_size + lvl2_size
    lvl1_prc_wted = bid_price1.values * ask_size1.values + ask_price1.values * bid_size1.values
    lvl2_prc_wted = bid_price2.values * ask_size2.values + ask_price2.values * bid_size2.values
    book['wap1'] = lvl1_prc_wted/lvl1_size
    #book['wap2'] = (bid_price1.values * (ask_size1.values + ask_size2.values) + ask_price1.values * (bid_size1.values + bid_size2.values))/(lvl1_size + lvl2_size)
    book['wap3'] = lvl2_prc_wted/lvl2_size
    book['wap4'] = (lvl1_prc_wted+lvl2_prc_wted)/(tot_size)
    book['wap_diff'] = np.abs(book['wap3'].values - book['wap1'].values)
    book[['log_return1', 'log_return3','log_return4']] = np.log(book[['wap1','wap3','wap4']]).groupby(['time_id']).diff()
    book['deep_hhi'] = np.select(condlist=[tot_size>0],
                                 choicelist=[(bid_size1.values+bid_size2.values)/tot_size**2+((ask_size1.values+ask_size2.values)/tot_size)**2],
                                 default=0.5
                                )
    book['hhi'] = np.select(condlist=[lvl1_size>0],
                                 choicelist=[(bid_size1.values/lvl1_size)**2 + (ask_size1.values/lvl1_size)**2],
                                 default=0.5
                                )
    
    book[['log_return1_abs','log_return3_abs','log_return4_abs','deep_imb_abs','bid_ask_imb_abs','hhi_abs','deep_hhi_abs']] = book[['log_return1', 'log_return3','log_return4','deep_imb','bid_ask_imb','hhi','deep_hhi']].abs()
    book[['log_return1_sqr','log_return3_sqr','log_return4_sqr','deep_imb_sqr','bid_ask_imb_sqr','hhi_sqr','deep_hhi_sqr']] = book[['log_return1','log_return3','log_return4','deep_imb','bid_ask_imb','hhi','deep_hhi']]**2
    book[['log_return1_thr','log_return3_thr','log_return4_thr','deep_imb_thr','bid_ask_imb_thr','hhi_thr','deep_hhi_thr']] = book[['log_return1_abs', 'log_return3_abs','log_return4_abs','deep_imb_abs','bid_ask_imb_abs','hhi_abs','deep_hhi_abs']]**3
    book[['log_return1_frt','log_return3_frt','log_return4_frt','deep_imb_frt','bid_ask_imb_frt','hhi_frt','deep_hhi_frt']] = book[['log_return1', 'log_return3','log_return4','deep_imb','bid_ask_imb','hhi','deep_hhi']]**4
    
    
    # use sum
    book['spread_wted'] = (book['time_diff'].values * book['spread_bps'].values)/600
    book['deep_spread_wted'] = (book['time_diff'].values * book['deep_spread_bps'].values)/600
    book['bid_spread_wted'] = (book['time_diff'].values * book['bid_spread_bps'].values)/600
    book['ask_spread_wted'] = (book['time_diff'].values * book['ask_spread_bps'].values)/600
    book["bid1_addition"] = np.select(condlist=[bid_price1.values == prev_bid_price1.values,
                                                bid_price1.values > prev_bid_price1.values],
                                      choicelist=[np.clip(bid_size1.values - prev_bid_size1.values,a_min=0,a_max=None),
                                                 bid_size1.values],
                                      default=0)
    book["ask1_addition"] = np.select(condlist=[ask_price1.values == prev_ask_price1.values,
                                                ask_price1.values < prev_ask_price1.values],
                                      choicelist=[np.clip(ask_size1.values - prev_ask_size1.values,a_min=0,a_max=None),
                                                 ask_size1.values],
                                      default=0)
    book["bid2_addition"] = np.select(condlist=[bid_price2.values == prev_bid_price2.values,
                                                bid_price2.values > prev_bid_price2.values],
                                      choicelist=[np.clip(bid_size2.values - prev_bid_size2.values,a_min=0,a_max=None),
                                                 bid_size2.values],
                                      default=0)
    book["ask2_addition"] = np.select(condlist=[ask_price2.values == prev_ask_price2.values,
                                                ask_price2.values < prev_ask_price2.values],
                                      choicelist=[np.clip(ask_size2.values - prev_ask_size2.values,a_min=0,a_max=None),
                                                 ask_size2.values],
                                      default=0)
    book["bid1_depletion"] = np.select(condlist=[bid_price1.values == prev_bid_price1.values,
                                                bid_price1.values < prev_bid_price1.values],
                                      choicelist=[np.clip(prev_bid_size1.values - bid_size1.values,a_min=0,a_max=None),
                                                 prev_bid_size1.values],
                                      default=0)
    book["ask1_depletion"] = np.select(condlist=[ask_price1.values == prev_ask_price1.values,
                                                ask_price1.values > prev_ask_price1.values],
                                      choicelist=[np.clip(prev_ask_size1.values-ask_size1.values,a_min=0,a_max=None),
                                                 prev_ask_size1.values],
                                      default=0)
    book["bid2_depletion"] = np.select(condlist=[bid_price2.values == prev_bid_price2.values,
                                                bid_price2.values < prev_bid_price2.values],
                                      choicelist=[np.clip(prev_bid_size2.values - bid_size2.values,a_min=0,a_max=None),
                                                 prev_bid_size2.values],
                                      default=0)
    book["ask2_depletion"] = np.select(condlist=[ask_price2.values == prev_ask_price2.values,
                                                ask_price2.values > prev_ask_price2.values],
                                      choicelist=[np.clip(prev_ask_size2.values-ask_size2.values,a_min=0,a_max=None),
                                                 prev_ask_size2.values],
                                      default=0)
    book['bid_size1_wted'] = (book['time_diff'].values * bid_size1.values)/600
    book['ask_size1_wted'] = (book['time_diff'].values * ask_size1.values)/600
    book['bid_size2_wted'] = (book['time_diff'].values * bid_size2.values)/600
    book['ask_size2_wted'] = (book['time_diff'].values * ask_size2.values)/600
    

    book_dict_wanted = {k: signal_operations[k] for k in signal_lst['book']}

    df_book_signal = book.groupby('time_id').agg(book_dict_wanted)
    df_book_signal = postprocess_book_signals(df_book_signal)
    
    df_book_signal_multi = []
    for thres in [240, 360, 480]:
        #df_book_part = book[(book['seconds_in_bucket'] >thres)&(book['seconds_in_bucket'] <=thres+120)].groupby('time_id').agg(book_dict_wanted)
        df_book_part = book[(book['seconds_in_bucket'] >= thres)].groupby('time_id').agg(book_dict_wanted)
        df_book_part = postprocess_book_signals(df_book_part)
        df_book_part.columns = [str(x)+"_"+str(thres) for x in df_book_part.columns]
        df_book_signal_multi.append(df_book_part)
    df_book_signal_multi = pd.concat(df_book_signal_multi, axis=1)
        
       
    if add_tif:
        df_tif = {}
        for thres in [240, 360, 480]:
            for side in ['bid','ask']:
                for lvl in ['1','2']:
                    #book_subset = book[(book['seconds_in_bucket'] >thres)&(book['seconds_in_bucket'] <=thres+120)][[side+'_price'+lvl,side+'_size'+lvl,'seconds_in_bucket']].reset_index()
                    book_subset = book[(book['seconds_in_bucket'] >= thres)][[side+'_price'+lvl,side+'_size'+lvl,'seconds_in_bucket']].reset_index()
                    df_lvl = book_subset.loc[(book_subset[[side+'_price'+lvl,side+'_size'+lvl]].shift(-1) != book_subset[[side+'_price'+lvl,side+'_size'+lvl]]).any(axis=1)]
                    time_diff = df_lvl['seconds_in_bucket'].values
                    time_ids = df_lvl['time_id'].values
                    ids, index = np.unique(time_ids, return_index=True)
                    splits = np.split(time_diff, index[1:])
                    tifs=[]
                    idx = []
                    for time_id, x in zip(ids.tolist(), splits):
                        #print(f"{side+lvl}: time id={time_id}, time diff = {x.mean()}")
                        tifs.append((np.diff(x).sum()+x[0])/len(x))
                        idx.append(time_id)
                    df_tif['tif_'+side+lvl+'_'+str(thres)] = pd.Series(tifs, index=idx)
        df_tif = 1/pd.concat(df_tif, axis=1)
    else:
        df_tif = pd.DataFrame()
        
        
    trade['size'] /= 100.0
    trade['mid'] = (trade['bid_price1'].values + trade['ask_price1'].values)/2.0
    price, mid, size, order_count = (trade[x] for x in ['price','mid','size','order_count'])
    trade['dollar_volume'] = size.values * price.values #/ tick_size * 0.01
    
    trade['time_diff'] = trade['seconds_in_bucket'].groupby('time_id').diff().fillna(trade['seconds_in_bucket'])
    trade['time_diff'] += 1
    
    trade['trade_cnt_rate'] = order_count.values / trade['time_diff'].values
    trade['trade_rate'] = size.values / trade['time_diff'].values
    
    prev_price = price.groupby("time_id").shift(1).values
    
    direction = np.select(condlist=[price.values > mid.values,
               price.values < mid.values,
               price.values > prev_price,
               price.values < prev_price],
               choicelist=[1,-1,1,-1],
               default=0)
    trade['direction_size'] = size.values * direction
    trade['direction_cnt'] = order_count.values * direction
    
    trade['log_return_trade'] = np.log(price).groupby('time_id').diff()
    
    trade[['log_return_trade_abs','direction_size_abs','direction_cnt_abs']] = trade[['log_return_trade','direction_size','direction_cnt']].abs()
    trade[['log_return_trade_sqr','direction_size_sqr','direction_cnt_sqr']] = trade[['log_return_trade','direction_size','direction_cnt']]**2
    trade[['log_return_trade_thr','direction_size_thr','direction_cnt_thr']] = trade[['log_return_trade_abs','direction_size_abs','direction_cnt_abs']]**3
    trade[['log_return_trade_frt','direction_size_frt','direction_cnt_frt']] = trade[['log_return_trade','direction_size','direction_cnt']]**4
    
    trade_thru = np.select(condlist=[price.values >= trade['ask_price1'].values,
                                             price.values <= trade['bid_price1'].values
                                             ],
                                    choicelist=[1,1],
                                    default=0
                                   )
    trade['trade_thru_size'] = size.values * trade_thru
    trade['trade_thru_cnt'] = order_count.values * trade_thru
      
    trade_dict_wanted = {k: signal_operations[k] for k in signal_lst['trade']}
    
    df_trade_signal = trade.groupby('time_id').agg(trade_dict_wanted)
    df_trade_signal = postprocess_trade_signals(df_trade_signal)
        
    df_trade_signal_multi = []
    for thres in [240, 360, 480]:
        #df_trade_part = trade[(trade['seconds_in_bucket'] >thres)&(trade['seconds_in_bucket'] <=thres+120)].groupby('time_id').agg(trade_dict_wanted)
        df_trade_part = trade[(trade['seconds_in_bucket'] >=thres)].groupby('time_id').agg(trade_dict_wanted)
        df_trade_part = postprocess_trade_signals(df_trade_part)
        df_trade_part.columns = [str(x)+"_"+str(thres) for x in df_trade_part.columns]
        df_trade_signal_multi.append(df_trade_part)
    df_trade_signal_multi = pd.concat(df_trade_signal_multi, axis=1)
    res = pd.concat([df_book_signal, df_tif, df_book_signal_multi, df_trade_signal, df_trade_signal_multi], axis=1)
    res.index.name = 'time_id'
    res['stock_id'] = stk_id
    return res.reset_index(drop=False)
def read_train_test():
    train = pd.read_csv('../input/optiver-realized-volatility-prediction/train.csv')
    test = pd.read_csv('../input/optiver-realized-volatility-prediction/test.csv')
    # Create a key to merge with book and trade data
    train['row_id'] = train['stock_id'].astype(str) + '-' + train['time_id'].astype(str)
    test['row_id'] = test['stock_id'].astype(str) + '-' + test['time_id'].astype(str)
    print(f'Our training set has {train.shape[0]} rows')
    return train, test
```

```python
def downsample_book(book, freq): # freq = 1s, 2s, ...; need 600 % freq == 0
    book_downsample = pd.merge(pd.DataFrame({"seconds_in_bucket": range(0, 601, freq)}, dtype = "int16"), pd.DataFrame({'time_id': book['time_id'].unique()}), how = 'cross')
    book_downsample = pd.merge_asof(book_downsample, book[['seconds_in_bucket','time_id','mid']].sort_values(['seconds_in_bucket']), on = 'seconds_in_bucket', by = 'time_id')
    return book_downsample.sort_values(["time_id", "seconds_in_bucket"]).reset_index(drop = True)
```

```python
def compute_downsampled_log_return_by_stock(stk_id, freq, mode = "train"):
    book = pd.read_parquet(f'/kaggle/input/optiver-realized-volatility-prediction/book_{mode}.parquet/stock_id={stk_id}')
    book["mid"] = (book["bid_price1"] + book["ask_price1"]) / 2.0
    book_downsample = downsample_book(book, freq)
    book_downsample['log_mid'] = np.log(book_downsample['mid'])
    book_downsample[stk_id] = book_downsample.groupby(["time_id"])["log_mid"].diff()
    #print(book_downsample)
    return book_downsample.set_index(['time_id','seconds_in_bucket'])[stk_id].dropna()

def compute_downsampled_log_return(stk_lst, freq, mode = "train"):
    res = Parallel(n_jobs = -1, verbose = 10)(delayed(partial(compute_downsampled_log_return_by_stock, freq = freq, mode = mode))(stk_id) for stk_id in stk_lst)
    return pd.concat(res, axis = 1).fillna(0)
```

```python
def compute_resid_vol(universe, freqs, mode='train'):
    res = {}
    for freq in freqs:
        market_book = compute_downsampled_log_return(universe, freq, mode)
        beta = {"mkt": dict()}
        reg_X = pd.concat({'mkt':market_book.mean(axis=1)}, axis=1)
        for stk_id in universe:
            ols = sm.OLS(market_book[stk_id], reg_X).fit()
            beta["mkt"][stk_id] = ols.params[0]
        beta = pd.DataFrame(beta)
        beta_ret = beta['mkt'] * pd.concat({x:reg_X['mkt'] for x in beta.index}, axis=1) #+ beta['idx_29'] * pd.concat({x:reg_X['idx_29'] for x in beta.index}, axis=1)+ beta['idx_43'] * pd.concat({x:reg_X['idx_43'] for x in beta.index}, axis=1) + beta['idx_69'] * pd.concat({x:reg_X['idx_69'] for x in beta.index}, axis=1)
        resid_ret = market_book - beta_ret
        mkt_ret_sqr, beta_ret_sqr, resid_ret_sqr = market_book**2, beta_ret**2, resid_ret**2
        res['mkt_vol_'+str(freq)] = np.sqrt(mkt_ret_sqr.groupby('time_id').sum().stack())
        res['beta_vol_'+str(freq)] = np.sqrt(beta_ret_sqr.groupby('time_id').sum().stack())
        res['resid_vol_'+str(freq)] = np.sqrt(resid_ret_sqr.groupby('time_id').sum().stack())
    res = pd.DataFrame(res)
    res.index.names = ['time_id','stock_id']
    return res.reset_index()
def get_time_stock(df):
    # Get realized volatility columns
    vol_cols = ['spread_wted_sum','log_return1_sqr_sum','log_return3_sqr_sum','log_return4_sqr_sum','dollar_volume_sum','log_return_trade_sqr_sum']
    sub_period_cols=[]
    for thres in [240, 360, 480]:
        sub_period_cols += [c+'_'+str(thres) for c in vol_cols]
    vol_cols += sub_period_cols
            

    # Group by the stock id
    df_stock_id = df.groupby(['stock_id'])[vol_cols].agg(['mean', 'std', 'max', 'min', ]).reset_index()
    # Rename columns joining suffix
    df_stock_id.columns = ['_'.join(col) for col in df_stock_id.columns]
    df_stock_id = df_stock_id.add_suffix('_' + 'stock')

    # Group by the stock id
    df_time_id = df.groupby(['time_id'])[vol_cols].agg(['mean', 'std', 'max', 'min', ]).reset_index()
    # Rename columns joining suffix
    df_time_id.columns = ['_'.join(col) for col in df_time_id.columns]
    df_time_id = df_time_id.add_suffix('_' + 'time')
    
    # group by cluster
    df['cluster'] = df['stock_id'].map(c1_dict)
    df_cluster = df.groupby(['time_id','cluster'])[['log_return1_sqr_sum','log_return3_sqr_sum','log_return4_sqr_sum','log_return_trade_sqr_sum']].agg(['mean']).reset_index()
    df_cluster.columns = ['_'.join(col) for col in df_cluster.columns]
    df_cluster = df_cluster.add_suffix('_' + 'cluster')
    #print(df_cluster)
    
    
    # Merge with original dataframe
    df = df.merge(df_stock_id, how = 'left', left_on = ['stock_id'], right_on = ['stock_id__stock'])
    df = df.merge(df_time_id, how = 'left', left_on = ['time_id'], right_on = ['time_id__time'])
    df = df.merge(df_cluster, how='left', left_on = ['time_id','cluster'], right_on=['time_id__cluster', 'cluster__cluster'])
    df.drop(['stock_id__stock', 'time_id__time', 'time_id__cluster', 'cluster__cluster'], axis = 1, inplace = True)
#     df.drop(['stock_id__stock', 'time_id__time'], axis = 1, inplace = True)
    #df.drop(['time_id__time'], axis = 1, inplace = True)
    # join index stocks 29 43 69
    for idx in [29, 43, 69]:
        df_idx = df[df['stock_id'] == idx][vol_cols+['time_id']]
        df_idx.columns = [col+'_stockid_'+str(idx) if col != 'time_id' else col for col in df_idx.columns]
        df = df.merge(df_idx, how='left',on=['time_id'])
    return df
```
