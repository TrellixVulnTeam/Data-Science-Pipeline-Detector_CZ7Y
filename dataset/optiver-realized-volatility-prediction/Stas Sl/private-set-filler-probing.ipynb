This notebook sets all targets to 0, except a single random one, which is set to 1. If this target is ignored in RMSPE calculations then you'll get a perfect RMSPE = 1.0, otherwise there will be RMSPE > 1.

After running this notebook 20 times, I got 8 out of 20 perfect 1.0 scores, so very roughly amount of data ignored during evaluation could be estimated between 1/2 and 1/3 of all ~150000 targets.

```python
import random
import pandas as pd

df = pd.read_csv('/kaggle/input/optiver-realized-volatility-prediction/test.csv')

df['row_id'] = df.stock_id.astype('str') + '-' + df.time_id.astype('str')
df = df[['row_id']]

df['target'] = 0
if len(df) > 3:
    df.iloc[random.randrange(len(df)), 1] = 1
    
df.to_csv('submission.csv', index=False)
```
