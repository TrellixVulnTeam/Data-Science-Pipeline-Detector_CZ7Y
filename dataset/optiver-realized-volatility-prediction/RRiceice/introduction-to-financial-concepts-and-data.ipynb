# 訂單簿
術語訂單簿是指按價格水平組織的特定證券或金融工具的買賣訂單電子清單。 訂單簿列出了在每個價格點被投標或提供的股票數量。

下面是一個股票的訂單簿的快照（我們稱之為股票 A），如您所見，所有預定的買單都在顯示為“買價”的訂單的左側，而所有預定的賣單都在右側 書的一側顯示為“提供/詢問”


![order_book_1](https://www.optiver.com/wp-content/uploads/2021/05/OrderBook3.png)

交易活躍的金融工具總是有一個密集的訂單簿（流動性賬簿）。 由於訂單簿數據是市場需求/供應的連續表示，因此它始終被視為市場研究的第一數據源。 

# 貿易
訂單簿是市場交易意圖的代表，但是市場需要以**相同**價格進行交易的買方和賣方。 因此，有時當有人想進行股票交易時，他們會查看訂單簿並找到有相反興趣的人進行交易。

例如，假設當您擁有上一段中的訂單簿時，您想購買 20 股 A 股。 然後，您需要找到一些願意通過總共賣出 20 股或更多股票來與您交易的人。 您從最低價開始查看賬簿的**報價**面：在 148 的水平上有 221 股賣出權益。您可以以 148 的價格**解除** 20 股並**保證* * 你的執行。 這將是您交易後產生的股票 A 的訂單簿：

![order_book2](https://www.optiver.com/wp-content/uploads/2021/05/OrderBook4.png)

在這種情況下，賣方賣出 20 股，買方買入 20 股，交易所將匹配買賣雙方的訂單，並向公眾廣播一條交易消息：

- 20 股 A 股股票在市場上以 148 的價格交易。

與訂單簿數據類似，交易數據對 Optiver 的數據科學家也極為重要，因為它反映了市場的活躍程度。 實際上，金融市場中一些常見的技術信號是直接從交易數據中衍生出來的，例如高低或總成交量。

# 做市與市場效率
想像一下，在另一天，股票 A 的訂單簿變得低於預期，而您再次想從所有有意的賣家那裡購買 20 股。 如您所見，這本書沒有上一本那麼稠密，可以說，與上一本相比，這本書**少了液體**。

![order_book_3](https://www.optiver.com/wp-content/uploads/2021/05/OrderBook5.png)

您可以在 148 處下單買入。但是，目前沒有人願意在 148 處向您出售，因此您的訂單將留在賬簿中，等待有人與之交易。 如果你運氣不好，價格會上漲，其他人開始出價 149，你根本買不到。 或者，您可以以 155 的價格下單買入。交易所會將此訂單與 149 的一股未平倉賣單匹配，因此您以 149 買入 1 手。同樣，您將以 150 的價格買入 12 股 , 和 7 股 151。與試圖以 148 買入相比，沒有沒有得到你想要的交易的風險，但你最終會以更高的價格買入。

您可以看到，在這樣一個低效率的市場中，交易是很困難的，因為交易會更加昂貴，如果您想要高質量地執行您的訂單，您需要應對更高的市場風險。 這就是為什麼投資者喜歡流動性，無論市場條件多麼極端，像 Optiver 這樣的做市商都會提供流動性。

做市商是一個公司或個人，他們積極地對證券的兩側市場進行報價，提供買入和賣出（稱為要價）以及每個市場的市場規模。 由於做市商會同時顯示買入和賣出訂單，因此有做市商存在的訂單簿將更具流動性，因此將為最終投資者提供更有效的市場，讓他們自由交易而無需擔心執行。

# 訂單簿統計
Optiver 數據科學家可以從原始訂單簿數據中獲得大量統計數據，以反映市場流動性和股票估值。 這些統計數據被證明是任何市場預測算法的基本輸入。 下面我們想列出一些常見的統計數據，以激勵 Kaggler 從訂單簿數據中挖掘更有價值的信號。

讓我們回到股票 A 的原始訂單簿

![order_book_1](https://www.optiver.com/wp-content/uploads/2021/05/OrderBook3.png)

**買賣差價**

由於不同股票在市場上的交易水平不同，我們採用最佳報價和最佳買入價的比率來計算買賣差價。

買賣差價的公式可以寫成以下形式：
$$BidAskSpread = BestOffer/BestBid -1$$

**加權平均價格(WAP)**

訂單簿也是股票估值的主要來源之一。 公平的賬面估值必須考慮兩個因素：訂單的水平和規模。 在本次比賽中，我們使用加權平均價格或 WAP 來計算瞬時股票估值併計算已實現的波動率作為我們的目標。

WAP的公式可以寫成如下，它考慮了頂層價格和數量信息：
$$ WAP = \frac{BidPrice_{1}*AskSize_{1} + AskPrice_{1}*BidSize_{1}}{BidSize_{1} + AskSize_{1}} $$

如您所見，如果兩本書分別在相同的價格水平上同時有出價和要價，那麼有更多報價的那本將產生較低的股票估值，因為書中有更多的意向賣家，而更多的賣家意味著 市場上更多的供應導致股票估值較低的事實。

請注意，在大多數情況下，在連續交易時間內，訂單簿不應出現買盤高於賣盤或賣盤的情況。 換句話說，最有可能的是，出價和要價永遠不應該是**交叉的。**

在本次比賽中，目標是從 WAP 構建的。 訂單簿快照的 WAP 為 147.5317797。

# 日誌返回(Log returns)

**我們如何比較昨天和今天的股票價格？**
最簡單的方法就是拿差價。這絕對是最直觀的方式，但是股票之間的價格差異並不總是具有可比性。例如，假設我們在股票 A 和股票 B 上都投資了 1000 美元，股票 A 從 100 美元漲到了 102 美元，股票 B 從 10 美元漲到了 11 美元。我們總共有 10 股 A ($1000 / $100=10) 產生的利潤為 10⋅($102−$100)=$20，總共 100 股 B 的股票收益為 $100。因此，股票 A 的價格漲幅更大，儘管股票 B 的漲幅要大得多。

我們可以通過將移動除以股票的起始價格來解決上述問題，有效地計算價格變化的百分比，也稱為股票收益。在我們的例子中，股票 A 的回報是 $102−$100$100=2% ，而股票 B 的回報是 $11−$10$10=10% 。股票回報與我們投資資本的百分比變化相吻合。

回報在金融中被廣泛使用，但是當需要一些數學建模時，首選對數回報。將 St 稱為股票 S 在時間 t 的價格，我們可以將 t1 和 t2 之間的對數回報定義為：
rt1,t2=log(St2St1)
 
通常，我們會查看固定時間間隔內的日誌返回，因此對於 10 分鐘的日誌返回，我們的意思是 rt=rt−10min,t。

日誌返回具有幾個優點，例如：

它們在時間 rt1,t2+rt2,t3=rt1,t3 上是相加的
常規回報不能低於 -100%，而對數回報沒有界限

# 已實現波動率
當我們交易期權時，我們模型的一個有價值的輸入是股票對數回報的標準差。在更長或更短的時間間隔內計算的對數回報的標準差會有所不同，因此通常將其標準化為 1 年期，並且年化標準差稱為波動率。

我們將計算所有連續賬面更新的對數回報，並將實際波動率 𝜎 定義為對數回報平方和的平方根。當我們交易期權時，我們模型的一個有價值的輸入是股票的標準差日誌返回。在更長或更短的時間間隔內計算的對數回報的標準差會有所不同，因此通常將其標準化為 1 年期，並且年化標準差稱為波動率。

我們將計算所有連續書籍更新的對數回報，並將實現的波動性 𝜎 定義為對數回報平方和的平方根。
$$
\sigma = \sqrt{\sum_{t}r_{t-1, t}^2}

我們希望保持定義盡可能簡單明了，這樣沒有金融知識的 Kaggler 就不會受到懲罰。 所以我們沒有年化波動率，我們假設對數回報的均值為 0。

# 比賽數據
在本次比賽中，Kagglers 的挑戰是從固定 10 分鐘窗口的賬面和交易數據中生成一系列短期信號，以預測下一個 10 分鐘窗口的實際波動率。 在 train/test.csv 中給出的目標可以通過相同的 **time_id** 和 **stock_id** 與原始訂單簿/交易數據鏈接。 特徵和目標窗口之間沒有重疊。

**在python中實現波動率計算**

在本次比賽中，我們的目標是預測短期實際波動率。 儘管無法共享目標的訂單簿和交易數據，但我們仍然可以使用我們提供的特徵數據呈現已實現的波動率計算。

由於已實現波動率是給定股票價格變化的統計量度，要計算價格變化，我們首先需要以固定間隔（1 秒）進行股票估值。 我們將使用我們提供的訂單簿數據的加權平均價格或 WAP。

```python
book_example['wap'] = (book_example['bid_price1'] * book_example['ask_size1'] +
                                book_example['ask_price1'] * book_example['bid_size1']) / (
                                       book_example['bid_size1']+ book_example['ask_size1'])
```

**股票的 WAP 如下圖**

```python
fig = px.line(book_example, x="seconds_in_bucket", y="wap", title='WAP of stock_id_0, time_id_5')
fig.show()
```

為了計算對數回報，我們可以簡單地取兩個連續的**WAP**之間的比率的對數。 第一行將有一個空返回，因為以前的書更新是未知的，因此空返回數據點將被刪除。

```python
def log_return(list_stock_prices):
    return np.log(list_stock_prices).diff() 
```

```python
book_example.loc[:,'log_return'] = log_return(book_example['wap'])
book_example = book_example[~book_example['log_return'].isnull()]
```

**讓我們繪製這個工具在這個時間段內的逐筆回報**

```python
fig = px.line(book_example, x="seconds_in_bucket", y="log_return", title='Log return of stock_id_0, time_id_5')
fig.show()
```

此特徵桶中股票 0 的實際成交量將為：

```python
def realized_volatility(series_log_return):
    return np.sqrt(np.sum(series_log_return**2))
realized_vol = realized_volatility(book_example['log_return'])
print(f'Realized volatility for stock_id 0 on time_id 5 is {realized_vol}')
```

# 幼稚的預測：使用過去實現的波動率作為目標

關於波動性的一個眾所周知的事實是它往往是自相關的。 我們可以使用這個屬性來實現一個簡單的模型，該模型只是通過使用最初 10 分鐘內已實現的波動率來“預測”已實現的波動率。

讓我們計算一下整個訓練集過去實現的波動率，看看單個天真的信號的預測能力。

```python
import os
from sklearn.metrics import r2_score
import glob
list_order_book_file_train = glob.glob('/kaggle/input/optiver-realized-volatility-prediction/book_train.parquet/*')
```

由於在本次比賽中數據是通過 stock_id 劃分的，以便 Kaggler 更好地管理內存，我們嘗試逐個計算股票的實際波動率並將它們合併到一個提交文件中。 請注意，如果我們加載單個文件，則作為分區列的股票 ID 不存在，因此我們將手動修復它。 我們將重用上一節中定義的對數返回和已實現的波動率函數。

```python
def realized_volatility_per_time_id(file_path, prediction_column_name):
    df_book_data = pd.read_parquet(file_path)
    df_book_data['wap'] =(df_book_data['bid_price1'] * df_book_data['ask_size1']+df_book_data['ask_price1'] * df_book_data['bid_size1'])  / (
                                      df_book_data['bid_size1']+ df_book_data[
                                  'ask_size1'])
    df_book_data['log_return'] = df_book_data.groupby(['time_id'])['wap'].apply(log_return)
    df_book_data = df_book_data[~df_book_data['log_return'].isnull()]
    df_realized_vol_per_stock =  pd.DataFrame(df_book_data.groupby(['time_id'])['log_return'].agg(realized_volatility)).reset_index()
    df_realized_vol_per_stock = df_realized_vol_per_stock.rename(columns = {'log_return':prediction_column_name})
    stock_id = file_path.split('=')[1]
    df_realized_vol_per_stock['row_id'] = df_realized_vol_per_stock['time_id'].apply(lambda x:f'{stock_id}-{x}')
    return df_realized_vol_per_stock[['row_id',prediction_column_name]]
```

循環遍歷每隻股票，我們可以得到過去實現的波動率作為對每隻股票的預測。

```python
def past_realized_volatility_per_stock(list_file,prediction_column_name):
    df_past_realized = pd.DataFrame()
    for file in list_file:
        df_past_realized = pd.concat([df_past_realized,
                                     realized_volatility_per_time_id(file,prediction_column_name)])
    return df_past_realized
df_past_realized_train = past_realized_volatility_per_stock(list_file=list_order_book_file_train,
                                                           prediction_column_name='pred')
```

讓我們將輸出數據幀與 train.csv 連接起來，看看訓練集上的樸素預測的性能。

```python
train['row_id'] = train['stock_id'].astype(str) + '-' + train['time_id'].astype(str)
train = train[['row_id','target']]
df_joined = train.merge(df_past_realized_train[['row_id','pred']], on = ['row_id'], how = 'left')
```

我們將通過兩個指標評估樸素預測結果：RMSPE 和 R 平方。 

```python
from sklearn.metrics import r2_score
def rmspe(y_true, y_pred):
    return  (np.sqrt(np.mean(np.square((y_true - y_pred) / y_true))))
R2 = round(r2_score(y_true = df_joined['target'], y_pred = df_joined['pred']),3)
RMSPE = round(rmspe(y_true = df_joined['target'], y_pred = df_joined['pred']),3)
print(f'Performance of the naive prediction: R2 score: {R2}, RMSPE: {RMSPE}')
```

樸素模型的表現並不令人驚訝，但作為基準，它是一個合理的開始。
