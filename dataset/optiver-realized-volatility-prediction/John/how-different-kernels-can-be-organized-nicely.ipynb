Ideas:

1. All the feature scripts are named `{feature_name}.py`. All features can be loaded by calling Loader method. like this:

```python
import feature_stat_2
train_df, test_df = feature_stat_2.Loader()
```

2. All the models are name {model_name}.py. Each model can be trained separately by calling `python nn_feature_stat.py -t -j2`, which will save all the trained model in `./models`, preproprocessing scalers in `./scalers` and OOF predictions in `./oof`. Inferencing by `python nn_feature_stat.py -j2`, which will produce `./predictions/nn_feature_stat_2_1.npy` as prediction output.
