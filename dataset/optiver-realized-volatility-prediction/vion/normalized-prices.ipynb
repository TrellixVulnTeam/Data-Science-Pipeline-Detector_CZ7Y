```python
import pyarrow.parquet as pq
import numpy as np
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (18,6)

data_path=Path('../input/optiver-realized-volatility-prediction')
!ls {data_path}
```

## Read order book data

```python
book_train = pq.read_table(data_path / 'book_train.parquet').to_pandas()
```

## Mean and Std. for a stock_id

```python
book_train.loc[book_train['stock_id']==0, 'bid_price1'].apply({'mean':'mean', 'std' : 'std'})
```

## Mean and Std all stocks

```python
book_train.groupby('stock_id')['bid_price1'].mean().hist()
```

```python
book_train.groupby('stock_id')['bid_price1'].std().hist()
```

## Price evolution: same time slot, all stocks

Plots used bid prices but similar figures can be obtained using ask prices.

```python
fig, ax = plt.subplots(figsize=(18,6))
book_train.loc[book_train['time_id']==5].groupby('stock_id')[['seconds_in_bucket','bid_price1']].plot(x='seconds_in_bucket', y='bid_price1', ax=ax, legend=False, alpha=0.4);
```

## Price evolution: same stock, all (several) time slots

Limited number of time slots in order to speed up plotting.

```python
fig, ax = plt.subplots(figsize=(18,6))
book_train.loc[(book_train['stock_id']==0) & (book_train['time_id']<1000)].groupby('time_id')[['seconds_in_bucket','bid_price1']].plot(x='seconds_in_bucket', y='bid_price1', ax=ax, legend=False, alpha=0.2);
```
