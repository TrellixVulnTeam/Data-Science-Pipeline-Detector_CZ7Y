```python
import numpy as np 
import pandas as pd
import glob
from tqdm.notebook import tqdm
```

```python
list_book_path = glob.glob('../input/optiver-realized-volatility-prediction/book_train.parquet/*')
list_trade_path = glob.glob('../input/optiver-realized-volatility-prediction/trade_train.parquet/*')
```

```python
# row_id of book data
set_row_id_book = []
for path in tqdm(list_book_path):
    stock_id = path.split('=')[1]
    df = pd.read_parquet(path)
    list_row_id = [f'{stock_id}-{time_id}' for time_id in df.time_id.unique()]
    set_row_id_book.extend(list_row_id)
set_row_id_book = set(set_row_id_book)
```

```python
# row_id of trade data
set_row_id_trade = []
for path in tqdm(list_trade_path):
    stock_id = path.split('=')[1]
    df = pd.read_parquet(path)
    list_row_id = [f'{stock_id}-{time_id}' for time_id in df.time_id.unique()]
    set_row_id_trade.extend(list_row_id)
set_row_id_trade = set(set_row_id_trade)
```
