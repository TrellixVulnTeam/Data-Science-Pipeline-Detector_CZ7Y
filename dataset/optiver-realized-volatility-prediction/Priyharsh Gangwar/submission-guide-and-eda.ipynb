```python
train_data = pd.read_csv("../input/optiver-realized-volatility-prediction/train.csv")
```

```python
# Number of unique stock
print(train_data.stock_id.nunique())
train_data.stock_id.unique()
#train_data['stock_id'].value_counts()
#All 112 stocks have 3830 values
```

**We get that total number of stocks are 112, and all of them have nearly 3830 values available in training data set, This is like a continuous time series with missing values in between, in reality the length of the time series is 32767. Thus we have to predict those missing values for each stocks**

```python
#Volatility variation for stock with stock_id == 1
train_data[train_data['stock_id']==1][['time_id','target']].set_index('time_id').plot(figsize=(10,5), grid=True)
```

```python
#Mean value of volatility for each stock as we can see some stocks have higher volatility and mean is around 0.004
train_data.groupby(['stock_id']).mean().reset_index()[['stock_id','target']].set_index('stock_id').plot()
```

```python
train_data[train_data['stock_id']==0]
```

These are the mean values of each stock when we take all those ~3800 values

```python
#mean value for each stock
train_data.groupby(['stock_id']).median().reset_index()
```

```python
#For basic model we will just predict this mean value of corresponding stock irrespective of timestamp
ser=train_data.groupby(['stock_id']).median()['target']
train_data.groupby(['stock_id']).median()['target']
```

```python
test_df=pd.read_csv('../input/optiver-realized-volatility-prediction/test.csv')
```

```python
test_df['target']=test_df.stock_id.apply(lambda x: ser[x])
```
