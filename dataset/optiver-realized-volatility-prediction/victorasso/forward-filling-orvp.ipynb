```python
import glob
import pyarrow
import numpy as np
import pandas as pd
from tqdm import tqdm
```

```python
BASE = '../input/optiver-realized-volatility-prediction'
```

```python
def ffill(data_df): 
    data_df=data_df.set_index(['time_id', 'seconds_in_bucket'])
    data_df = data_df.reindex(pd.MultiIndex.from_product([data_df.index.levels[0], np.arange(0,600)], names = ['time_id', 'seconds_in_bucket']), method='ffill')
    return data_df.reset_index()
```

```python
!mkdir book_train_ffill.parquet
```

```python
book_files = glob.glob(f'{BASE}/book_train.parquet/stock_id=*/*.parquet')

for file in tqdm(book_files):
    book_df = pd.read_parquet(file)
    ffill_df = ffill(book_df)
    
    file_name = f'./book_train_ffill.parquet/' + file.split('/')[-2]
    save_df = pyarrow.Table.from_pandas(ffill_df)
    pyarrow.parquet.write_to_dataset(save_df, root_path=file_name)
```
