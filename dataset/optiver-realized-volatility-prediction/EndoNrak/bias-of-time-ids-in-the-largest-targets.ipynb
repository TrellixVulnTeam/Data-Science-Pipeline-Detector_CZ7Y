```python
train = pd.read_csv("../input/optiver-realized-volatility-prediction/train.csv")
stock_ids = train["stock_id"].unique()
print(f"there are {len(stock_ids)} stock_ids in train.csv")
```

```python
all_time_id_series = pd.Series()
for stock_id in stock_ids:
    stock_i = train.query('stock_id==@stock_id', engine='python')
    all_targets = stock_i['target']
    all_time_ids = stock_i['time_id']
    all_time_id_series = pd.concat((all_time_id_series, all_time_ids))

all_time_id_counts = all_time_id_series.value_counts()
print(all_time_id_counts)
plt.bar(all_time_id_counts.index, all_time_id_counts, width=10)
plt.xlabel("time_id")
plt.ylabel("counts (in whole data)")
plt.show()
```

In train.csv, there are 3830 unique time_ids,
and each time_id has 111 or 112 stock_ids (almost all).
We can say, train_csv data is not biased.

```python
largest_n_target_number = 5
largest_n_time_id_series = pd.Series()
for stock_id in stock_ids:
    stock_i = train.query('stock_id==@stock_id', engine='python')
    stock_i = stock_i.sort_values('target', ascending=False).reset_index()
    largest_n_targets = stock_i.iloc[:largest_n_target_number]['target']
    largest_n_time_ids = stock_i.iloc[:largest_n_target_number]['time_id']
    largest_n_time_id_series = pd.concat((largest_n_time_id_series, largest_n_time_ids))

largest_n_time_id_counts = largest_n_time_id_series.value_counts()
print(largest_n_time_id_counts)
plt.bar(largest_n_time_id_counts.index, largest_n_time_id_counts, width=100)
plt.xlabel("time_id")
plt.ylabel(f"counts (in the largest {largest_n_target_number} targets)")
plt.show()
```

The counts of some time_ids are over 20 counts, and time_id=19725 is outstanding!  

I guess, in these time_ids, financial shocks happened and volatilities notably increased regardless of stock_id.

Financial shock may be very difficult to predict, I hope it wouldn't happen in evaluation three months. 

```python
smallest_n_target_number = 5
smallest_n_time_id_series = pd.Series()
for stock_id in stock_ids:
    stock_i = train.query('stock_id==@stock_id', engine='python')
    stock_i = stock_i.sort_values('target', ascending=False).reset_index()
    smallest_n_targets = stock_i.iloc[smallest_n_target_number:]['target']
    smallest_n_time_ids = stock_i.iloc[smallest_n_target_number:]['time_id']
    smallest_n_time_id_series = pd.concat((smallest_n_time_id_series, smallest_n_time_ids))

smallest_n_time_id_counts = smallest_n_time_id_series.value_counts()
print(smallest_n_time_id_counts)

plt.bar(smallest_n_time_id_counts.index, smallest_n_time_id_counts, width=5)
plt.xlabel("time_id")
plt.ylabel(f"counts (in the smallest {smallest_n_target_number} targets)")
plt.show()
```

In case the smallest volatility, there seems to be no outstanding time_id.
