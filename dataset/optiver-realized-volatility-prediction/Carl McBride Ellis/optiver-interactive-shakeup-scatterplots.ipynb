### [Optiver Realized Volatility Prediction competition](https://www.kaggle.com/c/optiver-realized-volatility-prediction/): Interactive shakeup scatterplots
####  The "Top 20"
Note: The entire merged scores dataframe can be found as a `csv` file in the **Data** section this notebook.

Note: The team names are those as of September 2021, and may have been changed since.


```python
import numpy  as np 
import pandas as pd
pd.set_option('display.max_columns', None)
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
sns.set(style="darkgrid")
sns.set_palette("bright")

csv_0 = pd.read_csv('../input/optiver-public-leaderboard-files/optiver-realized-volatility-prediction-publicleaderboard_00.csv',index_col=0)
csv_1 = pd.read_csv('../input/optiver-public-leaderboard-files/optiver-realized-volatility-prediction-publicleaderboard_01.csv',index_col=0)
csv_2 = pd.read_csv('../input/optiver-public-leaderboard-files/optiver-realized-volatility-prediction-publicleaderboard_02_v2.csv',index_col=0)
csv_3 = pd.read_csv('../input/optiver-public-leaderboard-files/optiver-realized-volatility-prediction-publicleaderboard_03.csv',index_col=0)
csv_4 = pd.read_csv('../input/optiver-public-leaderboard-files/optiver-realized-volatility-prediction-publicleaderboard_04.csv',index_col=0)
csv_5 = pd.read_csv('../input/optiver-public-leaderboard-files/optiver-realized-volatility-prediction-publicleaderboard_05.csv',index_col=0)
csv_6 = pd.read_csv('../input/optiver-public-leaderboard-files/optiver-realized-volatility-prediction-publicleaderboard_06.csv',index_col=0)
csv_7 = pd.read_csv('../input/optiver-public-leaderboard-files/PrivateLeaderboardRank.csv',index_col=0)

# create the positions
csv_0.insert(loc=2, column='position 0', value=np.arange(start=1, stop=(len(csv_0)+1) ))
csv_1.insert(loc=2, column='position 1', value=np.arange(start=1, stop=(len(csv_1)+1) ))
csv_2.insert(loc=2, column='position 2', value=np.arange(start=1, stop=(len(csv_2)+1) ))
csv_3.insert(loc=2, column='position 3', value=np.arange(start=1, stop=(len(csv_3)+1) ))
csv_4.insert(loc=2, column='position 4', value=np.arange(start=1, stop=(len(csv_4)+1) ))
csv_5.insert(loc=2, column='position 5', value=np.arange(start=1, stop=(len(csv_5)+1) ))
csv_6.insert(loc=2, column='position 6', value=np.arange(start=1, stop=(len(csv_6)+1) ))
csv_7.rename(columns = {'PrivateLeaderboardRank':'position 7'}, inplace = True)

csv_0 = csv_0.drop([           'SubmissionDate'], axis='columns')
csv_1 = csv_1.drop(['TeamName','SubmissionDate'], axis='columns')
csv_2 = csv_2.drop(['TeamName','SubmissionDate'], axis='columns')
csv_3 = csv_3.drop(['TeamName','SubmissionDate'], axis='columns')
csv_4 = csv_4.drop(['TeamName','SubmissionDate'], axis='columns')
csv_5 = csv_5.drop(['TeamName','SubmissionDate'], axis='columns')
csv_6 = csv_6.drop(['TeamName','SubmissionDate'], axis='columns')
csv_7 = csv_7.drop(['TeamName'], axis='columns')

csv_0.rename(columns={"Score":"score 0"},inplace=True)
csv_1.rename(columns={"Score":"score 1"},inplace=True)
csv_2.rename(columns={"Score":"score 2"},inplace=True)
csv_3.rename(columns={"Score":"score 3"},inplace=True)
csv_4.rename(columns={"Score":"score 4"},inplace=True)
csv_5.rename(columns={"Score":"score 5"},inplace=True)
csv_6.rename(columns={"Score":"score 6"},inplace=True)
csv_7.rename(columns={"PrivateLeaderboardScore":"score final"},inplace=True)

merged_df = csv_0.join(csv_1).join(csv_2).join(csv_3).join(csv_4).join(csv_5).join(csv_6).join(csv_7)

# output the merged dataframe
merged_df.to_csv('merged_Optiver_scores.csv', index=True)

# Take a look at the top 20
pd.set_option('display.max_columns', None)
merged_df.set_index("TeamName").sort_values(by='score final', ascending=True).head(20)
```

### Interactive shakeup scatterplot: Positions
##### Shakeup #1

```python
# https://matplotlib.org/stable/gallery/color/named_colors.html
color_1 = 'blue'
color_2 = 'olive'
color_3 = 'orange'
color_4 = 'crimson'
color_5 = 'limegreen'
color_6 = 'red'
color_7 = 'teal'
color_8 = 'yellowgreen'

fig  = px.scatter(merged_df,x='position 0',y='position 1',hover_name='TeamName').update_traces(marker=dict(color=color_1))
fig.update_layout(
    title="The black line represents Δ=0.",
    xaxis_title="Old Leaderboard position",
    yaxis_title="New Leaderboard position",)
#fig.update_layout(xaxis = dict(range=[0.17,0.3]))
#fig.update_layout(yaxis = dict(range=[0.17,0.3]))
fig.add_shape(type="line",x0=0, y0=0, x1=4000, y1=4000,line=dict(color="black",width=1.5,))
fig.update_traces(mode='markers', marker_size=3)
fig.show();
```
