## bitとaskについて

## 株価収益率(PER)
> PERとはPrice Earnings Ratioの略で、株価が1株当たり純利益（EPS：Earnings Per Share）の何倍まで買われているか、すなわち1株当たり純利益の何倍の値段が付けられているかを見る投資尺度です。現在の株価が企業の利益水準に対して割高か割安かを判断する目安として利用されます。PERの数値は、低いほうが株価は割安と判断されます。なお、1株当たり純利益は当期の予想数値を用いるのが一般的です。

PER(倍) = 株価 / 1株あたりの純利益

## Log returns

元ネタ<br>
https://www.kaggle.com/jiashenliu/introduction-to-financial-concepts-and-data?scriptVersionId=67183666&cellId=19

昨日と今日の株価をどのように比較できますか？
価格の違いは必ずしも株式間で比較できるとは限りません

- 株式Aに1000ドル投資(1株100ドルの株を10株)
- 株式Bにも1000ドル投資(1株10ドルの株を100株)
- 株式Aは、100ドルから102ドルに移動
- 株式Bは、10ドルから11ドルに移動

上記の状態で利益を計算すると・・・

- 株Aは、10×（＄102- ＄100）= ＄20の利益
- 株Bは、100×（＄11- ＄10）= ＄100の利益

ここでのポイントは、株価の上昇は株式Aの方が大きかったが、動きは株式Bの方が比例してはるかに大きかった。


上記の問題は、移動を株式の開始価格で除算し、株価の変化率（株価収益率とも呼ばれる）を効果的に計算することで解決できます。

- 株Aのリターン （＄102- ＄100） / ＄100 ＝ 2％
- 株Bのリターン （＄11- ＄10） / ＄10 ＝ 10％

株式のリターンは、投資資本の変化率と一致します。

財務では返品が広く使用されていますが、数学的なモデリングが必要な場合は常に対数返品が推奨されます。 𝑆𝑡を時間𝑡での株式の価格𝑆と呼ぶと、𝑡1と𝑡2の間の対数リターンを次のように定義できます。
𝑟𝑡1、𝑡2= log（𝑆𝑡2𝑆𝑡1）
 
通常、一定の時間間隔でログリターンを確認するため、10分のログリターンでは、𝑟𝑡= 𝑟𝑡−10𝑚𝑖𝑛、𝑡を意味します。

ログの戻り値には、次のようないくつかの利点があります。

それらは時間の経過とともに相加的です𝑟𝑡1、𝑡2+𝑟𝑡2、𝑡3=𝑟𝑡1、𝑡3
ログの返品には制限がありませんが、通常の返品は-100％を下回ることはできません。



## Realized volatility(実現されたボラティリティ)
オプションを取引する場合、モデルへの貴重な入力は、株式ログリターンの標準偏差です。
標準偏差は、より長い間隔またはより短い間隔で計算された対数リターンでは異なります。このため、通常は1年の期間に正規化され、年換算の標準偏差はボラティリティと呼ばれます。

このコンテストでは、10分間の本のデータが提供され、次の10分間のボラティリティがどうなるかを予測するように求められます。ボラティリティは次のように測定されます。

連続するすべての書籍の更新について対数リターンを計算し、実現されたボラティリティ𝜎を対数リターンの合計の平方根として定義します。
$$
\sigma = \sqrt{\sum_{t}r_{t-1, t}^2}
$$
 
対数リターンを計算するための株価としてWAPを使用する場合。

財務知識のないカグラーが罰せられないように、定義をできるだけ単純かつ明確に保ちたいと考えています。したがって、ボラティリティを年換算しておらず、対数リターンの平均が0であると想定しています。




```python
#log_returnの式
def log_return(list_stock_prices):
    return np.log(list_stock_prices).diff() 
```
