## Plots how a team score changed throughout the leaderboard updates with medal boundries for reference

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Put your team name below and run all cells:

```python
TEAM_NAME = 'Slawek Biel'
```

```python
final_df = pd.read_csv('../input/optiver-public-leaderboard-files/PrivateLeaderboardRank.csv')
final_df = final_df.rename(columns={'PrivateLeaderboardScore':'Score'}).sort_values('Score')
final_df.to_csv('final-privateleaderboard.csv', index=False)


lbs = [
               '../input/optiver-public-leaderboard/optiver-realized-volatility-prediction-publicleaderboard.csv',
               '../input/optiver-public-leaderboard/optiver-realized-volatility-prediction-private14_20_21.csv', 
               '../input/optiver-public-leaderboard-files/optiver-realized-volatility-prediction-publicleaderboard_02_v2.csv',
               '../input/optiver-public-leaderboard-files/optiver-realized-volatility-prediction-publicleaderboard_03.csv',
               '../input/optiver-public-leaderboard-files/optiver-realized-volatility-prediction-publicleaderboard_04.csv',
               '../input/optiver-public-leaderboard-files/optiver-realized-volatility-prediction-publicleaderboard_05.csv',
               '../input/optiver-public-leaderboard-files/optiver-realized-volatility-prediction-publicleaderboard_06.csv',
               'final-privateleaderboard.csv']
names = ['public','v1', 'v2', 'v3', 'v4', 'v5', 'v6', 'final']
```

```python
gold_scores, silver_scores, bronze_scores = [],[],[]
my_scores = []
first_scores =[]
for lb_path in lbs:
    df = pd.read_csv(lb_path)
    last_gold = 10 + len(df)//500
    last_silver = len(df)//20
    last_bronze = len(df)//10
    first_scores.append(df.iloc[0].Score)
    gold_scores.append(df.iloc[last_gold-1].Score)
    silver_scores.append(df.iloc[last_silver-1].Score)
    bronze_scores.append(df.iloc[last_bronze-1].Score)
    my_scores.append(df.loc[df.TeamName==TEAM_NAME].Score)
```

```python
plt.figure(figsize=(20,20))
plt.plot(names,first_scores,color='green', label='first score')
plt.plot(names,gold_scores,color='yellow', label='gold score')
plt.plot(silver_scores, color='gray', label='silver score')
plt.plot(bronze_scores, color='brown', label='bronze score')
plt.plot(my_scores, color = 'blue', label='your score')
plt.fill_between(names, y1=gold_scores, y2=first_scores, alpha=.2, color='yellow')
plt.fill_between(names,y1=silver_scores, y2=gold_scores, alpha=0.2, color='gray')
plt.fill_between(names,y1=bronze_scores, y2=silver_scores, alpha=0.2, color='brown')
plt.gca().invert_yaxis()
plt.legend()
plt.title(TEAM_NAME)
plt.show()
```
