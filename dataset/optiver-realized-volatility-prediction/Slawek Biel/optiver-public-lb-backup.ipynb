```python
import pandas as pd
pd.options.display.max_rows = 4000
```

```python
def read_lb(fname):
    df = pd.read_csv(fname)
    df['place'] = range(1, len(df)+1)
    return df
```

```python
public_lb = read_lb('../input/optiver-public-leaderboard/optiver-realized-volatility-prediction-publicleaderboard.csv')

merged = public_lb[['TeamId', 'TeamName', 'place','Score']].rename(columns={'place':'place_public', 'Score':'score_public'})

private_lbs = ['../input/optiver-public-leaderboard/optiver-realized-volatility-prediction-private14_20_21.csv', 
               '../input/optiver-public-leaderboard/optiver-realized-volatility-prediction-private28-10-21.csv',
               '../input/optiver-public-leaderboard-files/optiver-realized-volatility-prediction-publicleaderboard_02_v2.csv',
               '../input/optiver-public-leaderboard-files/optiver-realized-volatility-prediction-publicleaderboard_03.csv',
               '../input/optiver-public-leaderboard-files/optiver-realized-volatility-prediction-publicleaderboard_04.csv',
               '../input/optiver-public-leaderboard-files/optiver-realized-volatility-prediction-publicleaderboard_05.csv']
names = ['w2','w4_noise', 'w4', 'w6', 'w8', 'w10']
for private_name, suff in zip(private_lbs, names):
    private = read_lb(private_name)[['TeamId', 'place', 'Score']]
    private = private.rename(columns={'place':f'place_{suff}', 'Score':f'score_{suff}'})
    merged = pd.merge(merged, private, on ='TeamId')
```
