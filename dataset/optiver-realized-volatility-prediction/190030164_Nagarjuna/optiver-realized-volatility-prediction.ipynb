```python
upper_limit = df.target.mean() + 4*df.target.std()
upper_limit
```

```python
lower_limit = df.target.mean() -4*df.target.std()
lower_limit
```

```python
df[(df.target>upper_limit) | (df.target<lower_limit)]
```

```python

df2 = df[(df.target<upper_limit) & (df.target>lower_limit)]
df2.head()
```

```python
df2.shape
```

```python
Q1 = df2.target.quantile(0.25)
Q3 = df2.target.quantile(0.75)
Q1, Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python
lower_limit = Q1 - 1.5*IQR
upper_limit = Q3 + 1.5*IQR
lower_limit, upper_limit
```

```python
df2[(df2.target<lower_limit)|(df2.target>upper_limit)]
```

```python
df3 = df2[(df2.target>lower_limit)&(df2.target<upper_limit)]
df3
```

```python
df3.shape
```

```python
y=df3.target
train=df3.drop('target',axis=1)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(train,y,test_size=0.25,random_state=2)
```

```python
from sklearn.metrics import mean_absolute_error
def mean(y,predit):
    mae=mean_absolute_error(y,predit)
    return(mae)
```

```python
model1=LinearRegression(fit_intercept=True,
    normalize=False,
    copy_X=True,
    n_jobs=1000).fit(xtrain, ytrain)
pred1=model1.predict(xtest)
mean(ytest,pred1)
```

```python
y_train=ytrain.astype('int')
from sklearn.naive_bayes import MultinomialNB
knn=MultinomialNB(alpha=111.0, fit_prior=True, class_prior=None).fit(xtrain, y_train)
pred1=model1.predict(xtest)
mean(ytest,pred1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5,weights='uniform',algorithm='auto',leaf_size=30,p=2,metric='minkowski',metric_params=None,n_jobs=100,).fit(xtrain, y_train)
pred1=model1.predict(xtest)
mean(ytest,pred1)
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=20,init='k-means++',
    n_init=10,
    max_iter=300,
    tol=0.0001,
    precompute_distances='deprecated',
    verbose=0,
    random_state=42,
    copy_x=True,
    n_jobs='deprecated',
    algorithm='auto',).fit(xtrain, ytrain)
pred1=model1.predict(xtest)
mean(ytest,pred1)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB(priors=None, var_smoothing=1e-09).fit(xtrain, y_train)
pred1=gnb.predict(xtest)
mean(ytest,pred1)
```

```python
from xgboost import XGBRegressor
#xgboost model
xgb_m = XGBRegressor(learning_rate=0.12358413,n_estimators=1000,random_state=4200 ,max_depth = 400, n_jobs=100)                        
xgb_m.fit(xtrain,ytrain)
predict_valid = xgb_m.predict(xtest)
    #test_predict = xgb_m.predict(xtest)
mean(ytest,predict_valid)
    
```

```python
from sklearn.linear_model import LogisticRegression
li=LinearRegression(fit_intercept=True,
    normalize=False,
    copy_X=True,
    n_jobs=1000).fit(xtrain, ytrain)
pred2=li.predict(xtest)
mean(pred2,ytest)
```

```python
from sklearn.tree import DecisionTreeRegressor
model3=DecisionTreeRegressor().fit(xtrain, ytrain)
pred3=model3.predict(xtest)
mean(pred3,ytest)
```

```python
from sklearn.ensemble import RandomForestRegressor
clf = RandomForestRegressor(n_estimators=100, n_jobs=10, verbose=10,max_depth=100,max_leaf_nodes=1000,min_samples_split=2,min_samples_leaf=100)
clf.fit(xtrain, ytrain)
predict=clf.predict(xtest)
mean(ytest,predict)
```
