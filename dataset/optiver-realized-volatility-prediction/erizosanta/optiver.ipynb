# オーダーブック
オーダーブックとは、特定の証券や金融商品に対する売買注文を価格水準ごとに整理した電子リストを指す。オーダーブックには、各価格帯で買われている株数や売られている株数が記載されています。

下の図は、ある銘柄（銘柄Aとします）のオーダーブックのスナップショットです。見てわかるように、意図した買い注文はすべてブックの左側に "bid "として表示され、意図した売り注文はすべてブックの右側に "offer/ask "として表示されています。

![order_book_1](https://www.optiver.com/wp-content/uploads/2021/05/OrderBook3.png)

活発に取引される金融商品には、常に密度の高いオーダーブック（A liquid book）が存在します。オーダーブックのデータは、市場の需要と供給を継続的に表しているため、市場調査のための第一のデータソースとして常に考えられています。

# 取引
オーダーブックは、市場での取引の意思を表すものですが、市場では、取引を成立させるために、**同じ**価格の買い手と売り手が必要です。そのため、ある銘柄を取引したいと思ったとき、オーダーブックを見て、反対の立場の人を探して取引をすることがあります。 

例えば、前項のオーダーブックを持っているときに、ある銘柄Aを20株買いたいとします。すると、合計20株以上の売りで自分に不利な取引をしてくれる人を何人か見つける必要があります。帳簿の**売り**側を安い価格から順に確認していくと、148円の水準で221株の売り気配があります。148の価格で20株を**持ち越す**ことができ、**約定を保証する**ことができます。これが取引後の銘柄Aのオーダーブックになります。

![order_book2](https://www.optiver.com/wp-content/uploads/2021/05/OrderBook4.png)

このケースでは、売り手が20株を売り、買い手が20株を買った場合、取引所は売り手と買い手の注文を照合し、1つの取引メッセージが公開されます。

- 株式Aの20株が148の価格で市場で取引されました。

オーダーブックデータと同様に、取引データもまた、市場の活性度を反映するものとして、Optiverのデータサイエンティストにとって非常に重要なものです。実際、金融市場でよく見られるテクニカル指標の中には、ハイ・ローや総取引量など、取引データから直接得られるものもあります。

# マーケットメイキングと市場効率
別の日に、銘柄Aのオーダーブックが以下の形になり、あなたは再び、すべての意図的な売り手から20株を買いたいと思ったとします。見ての通り、このブックは前のブックほど密度が高くなく、前のブックに比べて**流動性が低い**と言えるでしょう。

![order_book_3](https://www.optiver.com/wp-content/uploads/2021/05/OrderBook5.png)

あなたは148で買うという注文を出すことができました。しかし、148で売ってくれる人はいないので、あなたの注文はブックの中で誰かが反対売買をするのを待っている状態になります。もし運が悪ければ、価格が上がり、他の人が149で入札を始め、あなたは全く買えなくなってしまいます。一方、155で買うという注文を出すこともできます。取引所はこの注文を、149で1株の未決済の売り注文と照合しますので、あなたは149で1ロットを購入したことになります。同様に、価格150で12株、151で7株を買うことになります。148で買おうとした場合と比べて、希望通りの取引ができないというリスクはありませんが、結果的に高い価格で購入することになります。

このような非効率な市場では、取引コストが高くなり、質の高い注文の執行を望むのであれば、より高い市場リスクに対処しなければならないため、取引を行うことが困難であることがお分かりいただけると思います。だからこそ、投資家には流動性が求められ、どんなに極端な市場環境であっても流動性を提供することができるOptiverのようなマーケットメイカーがいます。

マーケットメイカーとは、ある証券の両建て相場を積極的に提示する企業または個人のことで、ビッド（買値）とオファー（売値）をそれぞれの市場規模とともに提示します。マーケットメイカーはビッドとオファーの両方の注文を表示するため、マーケットメイカーが存在するオーダーブックは流動性が高くなり、投資家が約定を気にすることなく自由に取引できる、より効率的な市場が提供されることになります。

# オーダーブック統計
Optiverのデータサイエンティストは、生のオーダーブックデータから、市場の流動性や株式の評価を反映した多くの統計データを導き出すことができます。これらの統計値は、あらゆる市場予測アルゴリズムの基本的なインプットであることが証明されています。以下では、Kagglerがオーダーブックデータからより価値のあるシグナルをマイニングするためのヒントとなるような、一般的な統計値をいくつか挙げてみたいと思います。

株式Aの最初のオーダーブックに戻ってみましょう。

![order_book_1](https://www.optiver.com/wp-content/uploads/2021/05/OrderBook3.png)

**ビッド/アスク・スプレッド**

市場では様々な銘柄が異なるレベルで取引されているため、売呼値と買呼値の比率からビッド・アスク・スプレッドを算出します。

ビッド・アスク・スプレッドの計算式は以下のようになります。
$$BidAskSpread = BestOffer/BestBid -1 = (BestOffer- BestBid)/BestBid $$

**加重平均価格**

オーダーブックは、株式評価の主要な情報源の一つでもあります。ブックベースの公正な評価は、注文のレベルとサイズという2つの要素を考慮しなければなりません。本コンペティションでは、WAP（加重平均価格）を用いて瞬間的な株価評価を算出し、実現ボラティリティを目標値として算出しました。

WAPの計算式は、トップレベルの価格と数量の情報を考慮して、以下のように書くことができます。

$$ WAP = \frac{BidPrice_{1}*AskSize_{1} + AskPrice_{1}*BidSize_{1}}{BidSize_{1} + AskSize_{1}} $$

ご覧のように、同じ価格帯で買値と売値があるオーダーブックが2つある場合、売値が多い方が低い株価評価となります。これは、そのオーダーブックに意図的な売り手が多いためであり、売り手が多いということは、市場での供給が多いという事実を意味し、結果として株価評価が低くなります。

ほとんどの場合、継続的な取引時間中、オーダーブックにはビッドオーダーがオファー（アスク）オーダーよりも高いというシナリオは存在しないはずです。言い換えれば、ビッドとアスクがクロスすることはないということです。

このコンペティションでは、ターゲットはWAPから構築されます。オーダーブックのスナップショットのWAPは147.5317797です。

# 対数リターン

**昨日と今日の株価を比較するには？**

最も簡単な方法は、単に差額を取ることです。これは間違いなく最も直感的な方法ですが、**価格差**は必ずしも銘柄間で比較できるものではありません。例えば、A株とB株にそれぞれ $\$$1000ずつ投資し、A株が$\$$100から$\$$102に、B株が$\$$10から$\$$11になったとしましょう。A銘柄は10株で$\$$10の利益、B銘柄は100株で$\$$100の利益となりました。つまり、値上がり幅は株式**A**の方が大きく、動きは株式Bの方が比例してはるかに大きかったのです。

この問題を解決するには、株価の変動を開始時の株価で割ることで、**株価の変動率**を算出します。今回の例では、銘柄Aのリターンは$\frac{\$102 - \$100 }{\$100} = 2\%$、銘柄Bのリターンは$\frac{\$11 - \$10 }{\$10} = 10\%$でした。株式のリターンは、投下資本の変化率と一致します。

リターンは金融の世界では広く使われているが、数学的なモデル化が必要な場合は**対数リターン**が好まれる。$S_t$を$t$時点での$S$銘柄の価格とすると、$t_1$と$t_2$の間の対数リターンを次のように定義することができる。
$$
r_{t_1, t_2} = \log \left( \frac{S_{t_2}}{S_{t_1}} \right)
$$
通常は、一定の時間間隔での対数リターンを見るので、10分足の対数リターンとは、$r_t = r_{t - 10 min, t}$を意味します。

対数リターンにはいくつかの利点があり、例えば次のようなものがある。
- 時間を超えて加法的である $r_{t_1, t_2} + r_{t_2, t_3} = r_{t_1, t_3}$
- 通常のリターンは-100%以下になることはないが、対数リターンは境界がない。

# 実現ボラティリティ
オプションを取引する際、モデルへの重要なインプットとなるのが、株式の対数リターンの標準偏差です。標準偏差は、対数リターンが長い間隔で計算された場合と短い間隔で計算された場合とで異なります。このため、通常は1年単位で標準化され、年単位の標準偏差は **ボラティリティ** と呼ばれます。

このコンペティションでは、10分間のオーダーブックデータが与えられ、その後の10分間のボラティリティを予測していただきます。ボラティリティは以下のように測定されます。

随時更新されるオーダーブックの対数リターンを計算し、対数リターンの二乗和の二乗根として**実現ボラティリティ$\sigma$**を定義します。
$$
\sigma = \sqrt{\sum_{t}r_{t-1, t}^2}
$$
ここで、対数リターンを計算するために株式の価格として **WAP** を使用しています。

金融の知識のないKagglerがペナルティを受けないように、定義はできるだけシンプルで明確にしたいと考えています。そのため、ボラティリティの年率化は行わず、対数リターンの平均値は0であると仮定しています。

# コンペティションのデータ
このコンペティションでは、Kagglerは固定された10分間のオーダーブック及び取引データから一連の短期指標を生成し、次の10分間の実現ボラティリティを予測することに挑戦します。train/test.csvで与えられるターゲットは、生のオーダーブック/取引データと同じ**time_id**と**stock_id**で紐づけすることができます。特徴量とターゲットに重複はありません。

コンペティションのデータはパーティション化されたparquetファイルで提供されることに注意してください。この[notebook](https://www.kaggle.com/sohier/working-with-parquet)にparquetファイルの扱い方のチュートリアルがあります。

データの最初の行では、time_id 5, stock_id 0の**ターゲット**の実現ボラティリティは0.004136であることを意味します。指標を構築するために、**特徴量**のオーダーブックと取引データを確認します。

**Pythonでの実現ボラティリティ計算**

このコンペティションでは、短期の実現ボラティリティを予測することが目標です。ターゲットのオーダーブックと取引データは共有できませんが、提供した特徴量を使って実現ボラティリティの計算を提示することができます。

実現ボラティリティは、ある銘柄の価格変化の統計的な指標であるため、価格変化を計算するためには、まず一定間隔（1秒）での株価評価が必要です。ここでは、提供したオーダーブックデータの加重平均価格（WAP）を使用します。

```python
book_example['wap'] = (book_example['bid_price1'] * book_example['ask_size1'] +
                                book_example['ask_price1'] * book_example['bid_size1']) / (
                                       book_example['bid_size1']+ book_example['ask_size1'])
```

**WAPは以下のように表示されます。**

```python
fig = px.line(book_example, x="seconds_in_bucket", y="wap", title='WAP of stock_id_0, time_id_5')
fig.show()
```

対数リターンを計算するため、単純に連続する2つの**WAP**間の**比率の対数**を取ります。最初の行は、以前のオーダーブックが不明であるため、空のリターンとなり、空のリターンのデータポイントは削除されます。

```python
def log_return(list_stock_prices):
    return np.log(list_stock_prices).diff() 
```

```python
book_example.loc[:,'log_return'] = log_return(book_example['wap'])
book_example = book_example[~book_example['log_return'].isnull()]
```

**対数リターンをプロットしてみましょう。**

```python
fig = px.line(book_example, x="seconds_in_bucket", y="log_return", title='Log return of stock_id_0, time_id_5')
fig.show()
```

この特徴量におけるstock id 0の実現ボラティリティは次のようになります。

```python
def realized_volatility(series_log_return):
    return np.sqrt(np.sum(series_log_return**2))
realized_vol = realized_volatility(book_example['log_return'])
print(f'Realized volatility for stock_id 0 on time_id 5 is {realized_vol}')
```

# 単純予測: 過去の実現ボラティリティを目標値とする

ボラティリティに関する一般的な事実は、自己相関する傾向があることです。この特性を利用して、最初の10分間の実現ボラティリティを使用して、実現ボラティリティーを「予測」するだけの単純なモデルを実装することができます。

トレーニングセット全体の過去の実現ボラティリティを計算し、単一の単純な指標がどの程度予測できるかを見てみましょう。

```python
import os
from sklearn.metrics import r2_score
import glob
list_order_book_file_train = glob.glob('/kaggle/input/optiver-realized-volatility-prediction/book_train.parquet/*')
```

本コンペティションでは、Kagglerがメモリをより効率的に管理できるように、データをstock_idで分割しているため、実現ボラティリティを銘柄ごとに計算し、1つの提出ファイルにまとめてみます。1つのファイルをロードすると、パーティションカラムとしてのストックIDが存在しないことに注意して、手動で修正します。前回のセッションで定義した対数リターン関数と実現ボラティリティ関数を再利用します。

```python
def realized_volatility_per_time_id(file_path, prediction_column_name):
    df_book_data = pd.read_parquet(file_path)
    df_book_data['wap'] =(df_book_data['bid_price1'] * df_book_data['ask_size1']+df_book_data['ask_price1'] * df_book_data['bid_size1'])  / (
                                      df_book_data['bid_size1']+ df_book_data[
                                  'ask_size1'])
    df_book_data['log_return'] = df_book_data.groupby(['time_id'])['wap'].apply(log_return)
    df_book_data = df_book_data[~df_book_data['log_return'].isnull()]
    df_realized_vol_per_stock =  pd.DataFrame(df_book_data.groupby(['time_id'])['log_return'].agg(realized_volatility)).reset_index()
    df_realized_vol_per_stock = df_realized_vol_per_stock.rename(columns = {'log_return':prediction_column_name})
    stock_id = file_path.split('=')[1]
    df_realized_vol_per_stock['row_id'] = df_realized_vol_per_stock['time_id'].apply(lambda x:f'{stock_id}-{x}')
    return df_realized_vol_per_stock[['row_id',prediction_column_name]]
```

各銘柄で繰り返すことで、各銘柄の過去の実現ボラティリティを予測値として得ることができます。

```python
def past_realized_volatility_per_stock(list_file,prediction_column_name):
    df_past_realized = pd.DataFrame()
    for file in list_file:
        df_past_realized = pd.concat([df_past_realized,
                                     realized_volatility_per_time_id(file,prediction_column_name)])
    return df_past_realized
df_past_realized_train = past_realized_volatility_per_stock(list_file=list_order_book_file_train,
                                                           prediction_column_name='pred')
```

出力されたデータフレームをtrain.csvと結合して、トレーニングセットでの単純な予測のパフォーマンスを見てみましょう。

```python
train['row_id'] = train['stock_id'].astype(str) + '-' + train['time_id'].astype(str)
train = train[['row_id','target']]
df_joined = train.merge(df_past_realized_train[['row_id','pred']], on = ['row_id'], how = 'left')
```

単純な予測結果を2つの指標で評価します。RMSPE(平均平方二乗誤差率)とR2(決定係数)です。

```python
from sklearn.metrics import r2_score
def rmspe(y_true, y_pred):
    return  (np.sqrt(np.mean(np.square((y_true - y_pred) / y_true))))
R2 = round(r2_score(y_true = df_joined['target'], y_pred = df_joined['pred']),3)
RMSPE = round(rmspe(y_true = df_joined['target'], y_pred = df_joined['pred']),3)
print(f'Performance of the naive prediction: R2 score: {R2}, RMSPE: {RMSPE}')
```

単純なモデルの性能は驚くべきものではありませんが、ベンチマークとしては妥当なスタートと言えるでしょう。
