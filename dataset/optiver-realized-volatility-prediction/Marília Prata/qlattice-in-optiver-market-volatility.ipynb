```python
train, test = feyn.tools.split(train, ratio=(1,1), random_state=42)
```

```python
ql = feyn.connect_qlattice()
```

```python
ql.reset(random_seed=0)
```

```python
models = ql.auto_run(train, output_name="target", kind="classification", criterion="aic",threads=10,loss_function='binary_cross_entropy')
```

Epoch no. 1 - Tried 2363 models - Best loss: 1.02e+01

Epoch no. 2 - Tried 4352 models - Best loss: 1.00e-07

Epoch no. 8 - Tried 15829 models - Best loss: 1.00e-07

Epoch no. 9 - Tried 17953 models - Best loss: 1.00e-07

#Evaluating models

```python
best = models[0]
```
