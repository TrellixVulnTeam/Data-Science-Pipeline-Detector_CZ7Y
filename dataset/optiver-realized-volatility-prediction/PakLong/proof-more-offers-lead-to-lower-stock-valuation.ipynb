Here is the mathematics proof to explain why if two books have both bid and ask offers on the same price level respectively, the one with more offers in place will generate a lower stock valuation.

$$WAP = \frac{BidPrice \cdot AskSize + AskPrice \cdot  BidSize}{BidSize+AskSize}$$

$$WAP = \frac{(BidPrice + \frac{AskPrice \cdot  BidSize}{AskSize})AskSize}{BidSize+AskSize}$$

$$WAP = \frac{BidPrice + \frac{AskPrice \cdot  BidSize}{AskSize}}{\frac{BidPrice }{AskSize}+1}$$

So when $AskSize$ approaches to  $inf$, $WAP$ converges to $BidPrice$ ,where $BidPrice$ is the lower boundary of $WAP$, since $\{WAP \in \mathbb{R} | BidPrice\leq WAP \leq AskSize\}$.
