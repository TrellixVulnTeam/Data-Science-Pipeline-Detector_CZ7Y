# Normalization

```python
F_1_7min = data.F_1_7.min()
F_1_12min = data.F_1_12.min()
F_1_13min = data.F_1_13.min()
F_3_19min = data.F_3_19.min()
F_3_21min = data.F_3_21.min()
F_4_2min = data.F_4_2.min()
F_4_3min = data.F_4_3.min()
F_4_8min = data.F_4_8.min()
F_4_9min = data.F_4_9.min()
F_4_10min = data.F_4_10.min()
F_4_14min = data.F_4_14.min()
data.F_1_7 = data.F_1_7 - F_1_7min
data.F_1_12 = data.F_1_12 - F_1_12min
data.F_1_13 = data.F_1_13 - F_1_13min
data.F_3_19 = data.F_3_19 - F_3_19min
data.F_3_21 = data.F_3_21 - F_3_21min
data.F_4_2 = data.F_4_2 - F_4_2min
data.F_4_3 = data.F_4_3 - F_4_3min
data.F_4_8 = data.F_4_8 - F_4_8min
data.F_4_9 = data.F_4_9 - F_4_9min
data.F_4_10 = data.F_4_10 - F_4_10min
data.F_4_14 = data.F_4_14 - F_4_14min
data.F_1_7 = np.power(data.F_1_7,2)
data.F_1_12 = np.power(data.F_1_12,2)
data.F_1_13 = np.power(data.F_1_13,2)
data.F_3_19 = np.power(data.F_3_19,2)
data.F_3_21 = np.power(data.F_3_21,2)
data.F_4_2 = np.power(data.F_4_2,3)
data.F_4_3 = np.power(data.F_4_3,3)
data.F_4_8 = np.power(data.F_4_8,3)
data.F_4_9 = np.power(data.F_4_9,3)
data.F_4_10 = np.power(data.F_4_10,3)
data.F_4_14 = np.power(data.F_4_14,3)
```
