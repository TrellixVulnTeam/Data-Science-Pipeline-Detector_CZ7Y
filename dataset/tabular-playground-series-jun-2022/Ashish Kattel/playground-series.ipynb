```python
from tqdm import tqdm
### using columns wise mean
def impute(df, cols):
    for c in tqdm(cols):
        mean = df[cols].mean()
        df[cols] = df[cols].fillna(mean)
        
    return df
        
cols = remove_row_id.columns
mean_df = impute(remove_row_id, cols)
print(mean_df.isnull().any().sum())
    
```

```python
col = sb["row-col"].apply(lambda x:x.split("-"))
```

```python
ans = []
j=0
for row in tqdm(mean_df.iterrows()):
    for i in col[j:]:
        if int(i[0]) == int(row[0]):
            answer = row[1].loc[i[1]]
            ans.append(answer)
        else:
            break
            
        j+=1
            
            

    
```

```python
### using mean row wise
```

```python
def row_mean(df):
    mean_row = df.median(axis=1)
    
    return mean_row
        
```

```python
mean = row_mean(remove_row_id)
mean
```

```python
col[0][0]
# mean[col[0][0]]
```

```python
def compute_mean(mean, cols):
    ans = []
    for col in cols:
        ans.append(mean[int(col[0][0])])
        
    return ans

ans = compute_mean(mean, col)
```

```python
submission = pd.DataFrame({
    
    "row-col": sb["row-col"].values,
    "value": ans
    
})
```
