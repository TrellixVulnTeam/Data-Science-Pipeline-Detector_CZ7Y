Problem statement

The June edition of the 2022 Tabular Playground series is all about data imputation. The dataset has similarities to the May 2022 Tabular Playground, except that there are no targets. Rather, there are missing data values in the dataset, and your task is to predict what these values should be.

Import libraries

```python
import pandas as pd
import numpy as np
```

Load and read datasets

```python
input_path = "../input/tabular-playground-series-jun-2022"

data = pd.read_csv("/".join([input_path, "data.csv"]), index_col='row_id')
submission = pd.read_csv("/".join([input_path, "sample_submission.csv"]), index_col='row-col')
```

Analyse data

```python
data.info()
```

Add row column to data

```python
data.insert(loc=0, column='row_col', value=np.arange(len(data)))
data
```

Loop through rows and columns

```python
null_values = []
rows = []
cols = []
locate = pd.DataFrame()

for row in range(len(data)):
  for col in data:
    if(pd.isna(data.loc[row,col])):
      null_values.append(data.loc[row, col])
      rows.append(data['row_col'][row])
      cols.append([col])

print("Length of null_values is: ", len(null_values))
```

Impute null values with mean

```python
for col in data:
    if data[col].dtype=="object":
        data[col].fillna("not listed", inplace=True)
    if data[col].dtype=="int":
        data[col].fillna(data[col].mode()[0], inplace=True)
    if data[col].dtype=='float':
        data[col].fillna(data[col].mean(), inplace=True)

data
```

Prepare imputed values

```python
locate['rows'] = rows
locate['cols'] = cols

imputed_value = []
imputed_rows = []

for i in range(len(locate)):
  df_rows = rows[i]
  df_cols = cols[i]
  imputed_rows.append(data.loc[df_rows]['row_col'])
  imputed_value.append(data.loc[df_rows, df_cols].values)
  
imputed_value = np.concatenate(imputed_value, axis=0 )
```

Submission

```python
submission['value'] = imputed_value
submission['row-col'] = submission.index
submission = submission[['row-col', 'value']]

submission.to_csv('submission.csv', index=False)
submission
```
