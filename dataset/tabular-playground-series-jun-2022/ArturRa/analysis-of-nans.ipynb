```python
data = pd.read_csv('/kaggle/input/tabular-playground-series-jun-2022/data.csv')
```

```python
f4NaNs = data.iloc[:,66:].isna().sum(axis=1)
f4NaNs
```

```python
print('Without NaNs:', len(f4NaNs[f4NaNs == 0]))
print('With 1 NaN:', len(f4NaNs[f4NaNs == 1]))
print('With 2 NaNs:', len(f4NaNs[f4NaNs == 2]))
print('With 3 NaNs:', len(f4NaNs[f4NaNs == 3]))
print('With 4 NaNs:', len(f4NaNs[f4NaNs == 4]))
print('With 5 NaNs:', len(f4NaNs[f4NaNs == 5]))
print('With 6 NaNs:', len(f4NaNs[f4NaNs == 6]))
```

```python
f4_col_NaNs = {}
sum = 0
for col in data.iloc[:,66:].columns:
    f4_col = data.iloc[:,66:].isna()[col]
    f4_col_NaNs[col] = sorted(list(set(f4NaNs[f4NaNs == 1].index) & set(f4_col[f4_col == 1].index)))
    print(col, ': ', len(f4_col_NaNs[col]), sep='')
    sum += len(f4_col_NaNs[col])
print(sum)
```

```python
f4_col_2NaNs = {}
sum = 0
for col1 in data.iloc[:,66:].columns:
    for col2 in data.iloc[:,66:].columns:
        if (int(col1[4:]) >= int(col2[4:])):
            continue
        key = str(col1)+str(col2)
        f4_col1 = data.iloc[:,66:].isna()[col1]
        f4_col2 = data.iloc[:,66:].isna()[col2]
        f4_col_2NaNs[key] = sorted(list(set(f4NaNs[f4NaNs == 2].index) & set(f4_col1[f4_col1 == 1].index) & set(f4_col2[f4_col2 == 1].index)))
        print(col1, ' and ', col2, ': ', len(f4_col_2NaNs[key]), sep='')
        sum += len(f4_col_2NaNs[key])
print(sum)
```

```python
f4_col_3NaNs = {}
sum = 0
for col1 in data.iloc[:,66:].columns:
    for col2 in data.iloc[:,66:].columns:
        for col3 in data.iloc[:,66:].columns:
            if ( int(col1[4:]) >= int(col2[4:]) or int(col2[4:]) >= int(col3[4:])):
                continue
            key = str(col1)+str(col2)+str(col3)
            f4_col1 = data.iloc[:,66:].isna()[col1]
            f4_col2 = data.iloc[:,66:].isna()[col2]
            f4_col3 = data.iloc[:,66:].isna()[col3]
            f4_col_3NaNs[key] = sorted(list(set(f4NaNs[f4NaNs == 3].index) & set(f4_col1[f4_col1 == 1].index) & set(f4_col2[f4_col2 == 1].index) & set(f4_col3[f4_col3 == 1].index)))
            print(col1, ' and ', col2, ' and ', col3, ': ', len(f4_col_3NaNs[key]), sep='')
            sum += len(f4_col_3NaNs[key])
print(sum)
```

```python
f4_col_4NaNs = {}
sum = 0
for col1 in data.iloc[:,66:].columns:
    for col2 in data.iloc[:,66:].columns:
        for col3 in data.iloc[:,66:].columns:
            for col4 in data.iloc[:,66:].columns:
                if ( int(col1[4:]) >= int(col2[4:]) or int(col2[4:]) >= int(col3[4:]) or int(col3[4:]) >= int(col4[4:])):
                    continue
                key = str(col1)+str(col2)+str(col3)+str(col4)
                f4_col1 = data.iloc[:,66:].isna()[col1]
                f4_col2 = data.iloc[:,66:].isna()[col2]
                f4_col3 = data.iloc[:,66:].isna()[col3]
                f4_col4 = data.iloc[:,66:].isna()[col4]
                f4_col_4NaNs[key] = sorted(list(set(f4NaNs[f4NaNs == 4].index) & set(f4_col1[f4_col1 == 1].index) & set(f4_col2[f4_col2 == 1].index) & set(f4_col3[f4_col3 == 1].index) & set(f4_col4[f4_col4 == 1].index)))
                if len(f4_col_4NaNs[key]) > 0:
                    print(col1, ' and ', col2, ' and ', col3, ' and ', col4, ': ', len(f4_col_4NaNs[key]), sep='')
                sum += len(f4_col_4NaNs[key])
print(sum)
```
