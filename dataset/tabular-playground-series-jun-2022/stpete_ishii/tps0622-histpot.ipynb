```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
```

```python
train=pd.read_csv('../input/tabular-playground-series-jun-2022/data.csv')
```

```python
display(train[0:3])
#display(train.info())
```

```python
cols=train.columns.tolist()
print(cols)
print(len(cols))
```

```python
fig, ax = plt.subplots(41,2,figsize=(15,90))
for i in tqdm(range(81)):
    r=i//2
    c=i%2
    sns.histplot(train[cols[i]], label=cols[i], ax=ax[r,c], color='C1',bins=40)
    ax[r,c].legend()
    ax[r,c].grid()
plt.show()
```
