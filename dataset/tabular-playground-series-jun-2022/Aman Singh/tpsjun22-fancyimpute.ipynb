```python
import pandas as pd
import numpy as np
from fancyimpute import SoftImpute, BiScaler
from tqdm import tqdm
import gc
```

```python
data = pd.read_csv('../input/tabular-playground-series-jun-2022/data.csv')
```

```python
corr = data.corr()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
correlation = corr
mask = np.triu(np.ones_like(correlation, dtype=np.bool))
plt.figure(figsize=(80, 50))
sns.heatmap(correlation,annot=True,mask=mask)
```

```python
import missingno as msno
msno.matrix(data.sample(1000))
```

```python
gc.collect()
```

```python
data_softimpute = data.copy()
data_softimpute[:] = BiScaler().fit_transform(data_softimpute.values)
data[:] = SoftImpute().fit_transform(data_softimpute)
```

```python
data
```

```python
sub = pd.read_csv('../input/tabular-playground-series-jun-2022/sample_submission.csv', index_col='row-col')
```

```python
%%time
for i in tqdm(sub.index):
    row = int(i.split('-')[0])
    col = i.split('-')[1]
    sub.loc[i, 'value'] = data.loc[row, col]
```

```python
sub.to_csv('sub_tps_jun22_softimpute_1.csv')
```
