```python
#distribution of data with rows containing null values removed
features_3 = data3.columns
i=0
while i < len(features_3):
    fig, axs = plt.subplots(1,5,figsize=(10, 3))
    sns.histplot(data=data3,x=data3[features_3[i]], kde=True, color="skyblue",ax=axs[0])
    axs[0].set_title(f'{features_3[i]}')
    sns.histplot(data=data3,x=data3[features_3[i+1]], kde=True, color="skyblue",ax=axs[1])
    axs[1].set_title(f'{features_3[i+1]}')
    sns.histplot(data=data3,x=data3[features_3[i+2]], kde=True, color="skyblue",ax=axs[2])
    axs[2].set_title(f'{features_3[i+2]}')
    sns.histplot(data=data3,x=data3[features_3[i+3]], kde=True, color="skyblue",ax=axs[3])
    axs[3].set_title(f'{features_3[i+3]}')
    sns.histplot(data=data3,x=data3[features_3[i+4]], kde=True, color="skyblue",ax=axs[4])
    axs[4].set_title(f'{features_3[i+4]}')

    i+=5
    plt.tight_layout()
    plt.show()
```

```python
features_3 = data3.columns
i=0
while i < len(features_3):
    fig, axs = plt.subplots(1,5,figsize=(10, 3))
    sns.boxplot(data=data3,x=data3[features_3[i]],color="skyblue",ax=axs[0])
    axs[0].set_title(f'{features_3[i]}')
    sns.boxplot(data=data3,x=data3[features_3[i+1]],color="skyblue",ax=axs[1])
    axs[1].set_title(f'{features_3[i+1]}')
    sns.boxplot(data=data3,x=data3[features_3[i+2]],color="skyblue",ax=axs[2])
    axs[2].set_title(f'{features_3[i+2]}')
    sns.boxplot(data=data3,x=data3[features_3[i+3]],color="skyblue",ax=axs[3])
    axs[3].set_title(f'{features_3[i+3]}')
    sns.boxplot(data=data3,x=data3[features_3[i+4]],color="skyblue",ax=axs[4])
    axs[4].set_title(f'{features_3[i+4]}')

    i+=5
    plt.tight_layout()
    plt.show()
```

```python
#kurtosis and skewness from data with rows containing null values removed
from scipy.stats import norm, kurtosis,skew
    
print(f'Kurtosis: {kurtosis(data3)}, Skewness: {skew(data3)}')
```

Most features contain a kurtosis value of less than 3 meaning that the data is mostly playkurtic in that the distribution is very narrow and produces less extreme tails than the normal distribution.

Most features also contain skew values close to zero indicating most features contain some level of skewness but nothing too extreme.
