# Impute Missing Values

A simple imputer will be created that will fill the missing values. 

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
imp_mean = IterativeImputer(max_iter=20, random_state=0)
```

```python
trans_data = imp_mean.fit_transform(data)
```

```python
df = pd.DataFrame(trans_data, columns = data.columns)
```

# Create Submission

```python
sample_df = pd.read_csv("../input/tabular-playground-series-jun-2022/sample_submission.csv")
```

```python
value_list = []
for idx, row in sample_df.iterrows():
    value_list.append(df.loc[int(row["row-col"].split("-")[0])][row["row-col"].split("-")[1]])
    if (idx + 1) % 100000 == 0:
        print(idx + 1, "done")
```

```python
sample_df["value"] = value_list
sample_df.to_csv("submission.csv", index=False)
```
