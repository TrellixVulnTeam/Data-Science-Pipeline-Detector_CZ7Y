```python
import pandas as pd
import numpy as np
from tqdm.notebook import tqdm

import sys
sys.path.append("kuma_utils/")
from kuma_utils.preprocessing.imputer import LGBMImputer
```

# Data loading ...

```python
input_path = "../input/tabular-playground-series-jun-2022"

data = pd.read_csv("/".join([input_path, "data.csv"]), index_col='row_id')
submission = pd.read_csv("/".join([input_path, "sample_submission.csv"]), index_col='row-col')
```

# LGBMImputer

```python
lgbm_imp = LGBMImputer(n_iter=100, verbose=True)

data = lgbm_imp.fit_transform(data)
```

# Submission

```python
for i in tqdm(submission.index):
    row = int(i.split('-')[0])
    col = i.split('-')[1]
    submission.loc[i, 'value'] = data.loc[row, col]

submission.to_csv("submission.csv")
submission
```
