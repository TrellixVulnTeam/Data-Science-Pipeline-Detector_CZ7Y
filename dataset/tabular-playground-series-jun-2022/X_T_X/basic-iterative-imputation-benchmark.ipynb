```python
import numpy as np
import pandas as pd
from tqdm import tqdm
```

```python
data = pd.read_csv('../input/tabular-playground-series-jun-2022/data.csv')
```

```python
cat_features = []
for c in data.columns:
    if c[0:3] == 'F_2':
        cat_features.append(c)
        
num_features=[]
for c in data.columns:
    if c not in cat_features:
        num_features.append(c)
```

```python
data = data.drop('row_id',axis=1)
```

```python
from sklearn.linear_model import BayesianRidge
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
imputer = IterativeImputer(estimator=BayesianRidge(), 
                           n_nearest_features=None, imputation_order='ascending')
imputer.fit(data)
```

```python
Xtrans = imputer.transform(data)
```

```python
df = pd.DataFrame(Xtrans, columns = data.columns)
```

```python
submission = pd.read_csv('../input/tabular-playground-series-jun-2022/sample_submission.csv', index_col='row-col')
```

```python
from tqdm import tqdm
for i in tqdm(submission.index):
    row = int(i.split('-')[0])
    col = i.split('-')[1]
    submission.loc[i, 'value'] = df.loc[row, col]

submission.to_csv('benchmark1.csv')
```
