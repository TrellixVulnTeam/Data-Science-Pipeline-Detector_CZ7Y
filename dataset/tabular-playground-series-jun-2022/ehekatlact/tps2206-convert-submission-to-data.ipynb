```python
import numpy as np
import pandas as pd
from collections import defaultdict
from tqdm.notebook import tqdm
```

```python
data_df = pd.read_csv('../input/tabular-playground-series-jun-2022/data.csv', index_col='row_id')
sub = pd.read_csv('../input/tabular-playground-series-jun-2022/sample_submission.csv', index_col='row-col')
```

```python
other_sub = pd.read_csv("../input/tps-jun-22-lama-lightautoml-loop/lightautoml.csv", index_col='row-col')
```

### convert to data

```python
# best_score_dic[col] = [(index, value)]
other_sub_dic = defaultdict(list)
for i in tqdm(other_sub.index):
    row = int(i.split('-')[0])
    col = i.split('-')[1]
    val = other_sub.loc[i, 'value']
    other_sub_dic[col].append((row, val))
```

```python
for col in tqdm(data_df.columns):
    if not other_sub_dic[col]:
        continue
    index_value_list = other_sub_dic[col]
    index_list = []
    value_list = []
    for index, value in index_value_list:
        index_list.append(index)
        value_list.append(value)
    data_df[col].loc[index_list] = value_list
```

```python
data_df.to_pickle("data_df.pkl")
```

### make sub

```python
for i in tqdm(sub.index):
    row = int(i.split('-')[0])
    col = i.split('-')[1]
    sub.loc[i, 'value'] = data_df.loc[row, col]
```
