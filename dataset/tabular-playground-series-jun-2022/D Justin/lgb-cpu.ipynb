```python
import pandas as pd
import numpy as np
data = pd.read_csv('../input/tabular-playground-series-jun-2022/data.csv')
submission = pd.read_csv('../input/tabular-playground-series-jun-2022/sample_submission.csv')
```

```python
data_completed = pd.DataFrame()
```

```python
from catboost import CatBoostRegressor

index_list = data.columns.to_list()

for index in index_list[11:21]:
    if data[index].isnull().any():
        print(True)

        missing_list = list(np.where(data[index].isnull())[0])
        no_missing_list = list(np.where(data[index].isnull() == False)[0])
        train = data.iloc[no_missing_list,]
        test = data.iloc[missing_list,]
        X = train.drop([index,'row_id'],axis=1)
        y = train[index]
        X_test = test.drop([index,'row_id'],axis=1)
        model = CatBoostRegressor(iterations=20000,random_state=0,task_tpye='GPU')
        model.fit(X,y)
        y_predict = model.predict(X_test)
        data_all = data[index]
        data_all.iloc[missing_list,] = y_predict
        
        data_completed = pd.concat([data_completed,data_all],axis=1)
    else:
        data_completed = pd.concat([data_completed,data[index]],axis=1)
        print(False)
```
