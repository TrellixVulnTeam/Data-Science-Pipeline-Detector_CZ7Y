# TPS-June-2022: Create Pickle Dataset

```python
import glob
import numpy as np
import pandas as pd
data = pd.read_csv(glob.glob("../input/**/data.csv")[0])
for column in data.columns:
    if data[column].dtypes == "float64":
        data[column] = data[column].astype(np.float16)
    if data[column].dtypes == "int64":
        if data[column].min() >= -128 and data[column].max() <= 127:
            data[column] = data[column].astype(np.int8)
        elif data[column].min() >= -32768 and data[column].max() <= 32767:
            data[column] = data[column].astype(np.int16)
        elif data[column].min() >= -2147483648 and data[column].max() <= 2147483647:
            data[column] = data[column].astype(np.int32)
data.to_pickle("data.pkl")
data.info()
```
