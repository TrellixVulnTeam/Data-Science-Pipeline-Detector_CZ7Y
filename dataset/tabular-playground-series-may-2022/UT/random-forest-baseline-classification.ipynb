```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import numpy as np
import os

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
train_df = pd.read_csv("/kaggle/input/tabular-playground-series-may-2022/train.csv")
test_df = pd.read_csv("/kaggle/input/tabular-playground-series-may-2022/test.csv")
sample_submission_df = pd.read_csv("/kaggle/input/tabular-playground-series-may-2022/sample_submission.csv")
```

```python
x_train, x_test, y_train, y_test = train_test_split(train_df.drop(["id","target","f_27"], axis=1), train_df.target)
```

```python
%%time
model = RandomForestClassifier(n_jobs=-1)
model.fit(x_train, y_train)
```

```python
prediction = model.predict(test_df.drop(["id","f_27"],axis=1))
```

```python
prediction_df = test_df[["id"]]
prediction_df["target"] = prediction.astype(int)
```
