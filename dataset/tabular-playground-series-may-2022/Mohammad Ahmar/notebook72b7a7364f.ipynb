```python
#Notes before EDA:
#No missing values, no categorical features(except f_27) 
#Should find correlations between features and target feature
```

```python
plt.rcParams['figure.figsize'] = (20,20) 
fig,ax = plt.subplots(5,6)
fig.text(0.45,1, "BASIC EDA", {'size': 35})
i=0
j=0
for col in train.columns:
    if col not in ['f_27','target']:
        ax[j,i].hist(train[col], bins=80)
        ax[j,i].set_title(col, {'size':15})
        if i==5:
            i=0
            j+=1
        else:
            i+=1
plt.show()
```

```python
train['f_27'].value_counts()[:50].plot(kind='bar')
plt.title("f_27", {'size':20,'weight':'bold'})
plt.show()
```

```python
plt.figure(figsize=(30,30))
sns.heatmap(train.corr(), annot=True)
plt.title("Feature correlations",{'size':'35'})
plt.show()
```

```python
#Post EDA assumptions:
#Not all features are continous some are discrete
#f_21 and f_18 have the most correlation
#Maybe try dropping f_27(categorical feature) because of too many unique values and low target correlation
```
