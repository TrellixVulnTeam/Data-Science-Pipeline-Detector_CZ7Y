# 2. Loading the Model Results, Submissions from the Notebooks

```python
%%time
grv_vsn = pd.read_csv('../input/classification-with-grv-vsn/my_submission_051322.csv')
gbdt = pd.read_csv('../input/tps-may22-eda-gbdt/my_submission_043022.csv')
nn = pd.read_csv('../input/tps-may22-eda-neuronal-nets/my_submission_052222_v5.csv')
```

# 4. Creating a Weight Array and Blended Model

```python
%%time
weights = [0.01, 0.04, 0.95]
ensamble = weights[0] * grv_vsn + weights[1] * gbdt + weights[2] * nn
```

# 5. Creating Ensamble Model Submission

```python
%%time
sub = pd.read_csv('/kaggle/input/tabular-playground-series-may-2022/sample_submission.csv')

# Populated the prediction on the submission dataset and creates an output file
sub['target'] = ensamble['target']
sub.to_csv('my_ensamble_052222.csv', index = False)
```
