```python
import pandas as pd
import numpy as np
from sklearn.metrics import roc_auc_score
```

When ensembling the results of multiple models, the ensemble may not be properly effective if there are differences in the scales of each model.

Please upvote, if this helps.

```python
target = np.array([1.0, 1.0, 0.0, 0.0, 1.0])

pred1 = np.array([10, 3, 7, 0, 40]) 
pred2 = np.array([1.0, 0.9, 0.4, 0.4, 0.1])
```

```python
print('pred1 score: ', roc_auc_score(target, pred1))
print('pred2 score: ', roc_auc_score(target, pred2))
```

```python
ensamble = (pred1 + pred2) / 2
print('ensembled array: ', ensamble)
print('ensembled score: ', roc_auc_score(target, ensamble))
```

```python
from sklearn.preprocessing import MinMaxScaler

MMscaler = MinMaxScaler()

pred1_mm = MMscaler.fit_transform(pred1.reshape(-1,1)).reshape(-1)
pred2_mm = MMscaler.fit_transform(pred2.reshape(-1,1)).reshape(-1)

print('pred1 array after　scaling: ', pred1_mm)
print('pred1 score after scaling: ', roc_auc_score(target, pred1_mm))
print('')
print('pred2 array after　scaling: ', pred2_mm)
print('pred2 score after　scaling: ', roc_auc_score(target, pred2_mm))
```

As, you can see above, the score remains the same even if we scale the predictions


```python
ensamble_mm = (pred1_mm + pred2_mm) / 2

print('ensembled array　after　scaling: ', ensamble_mm)
print('ensembled score　after　scaling: ', roc_auc_score(target, ensamble_mm))
```

Well this turned out to be quite the extreme example, but i think you get the gist of it.
