```python
import pandas as pd
```

```python
subs = [
    pd.read_csv('../input/tps-may-2022/submission-99822.csv'), #99822
    pd.read_csv('../input/tps-may-2022/submission-99821.csv'), #99821    
    pd.read_csv('../input/tps-may-2022/submission-99814.csv') #99814
    ]
sub = subs[0].copy()
```

```python
W = [0.8, 0.8, 0.6]

total = sum(W)
W = [w / total for w in W]

state = 0
for i in range(len(W)):
    state += W[i] * subs[i]['target']
sub['target'] = state
```

```python
state_9821 = subs[0]['target']
state_9822 = sub['target']
diff = sub['target'] - subs[0]['target']

pd.DataFrame(data={'state_9821': state_9821, 'state_9822': state_9822, 'difference': diff})
```

```python
sub.to_csv('submission.csv', index=False)
```
