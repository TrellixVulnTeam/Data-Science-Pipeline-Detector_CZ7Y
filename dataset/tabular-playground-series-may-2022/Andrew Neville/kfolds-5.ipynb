```python
df_train = pd.read_csv('/kaggle/input/tabular-playground-series-may-2022/train.csv')
df_test = pd.read_csv('/kaggle/input/tabular-playground-series-may-2022/test.csv')
df_sampleSubmission = pd.read_csv('/kaggle/input/tabular-playground-series-may-2022/sample_submission.csv')
```

```python
# set up a 'folds' column
# populate with -1 for now
df_train['kfolds'] = -1
```

```python
# now populate the 'folds' column with a fold identifier
# try 5 folds for now
fold_no = 5
kf = model_selection.KFold(n_splits=fold_no, shuffle=True, random_state=0)
for fold, (train_indicies, valid_indicies) in enumerate(kf.split(X=df_train)):
    df_train.loc[valid_indicies, "kfolds"] = fold
```
