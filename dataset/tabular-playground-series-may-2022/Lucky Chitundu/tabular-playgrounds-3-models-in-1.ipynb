**PREPROCESSING**

```python
y = df["target"]
x = df.drop(["id","target","f_27"],axis=1)
tst.drop(columns=["id","f_27"],inplace=True)
```

**SPLITTING**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

**MODEL CONSTRUCTION**

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state=0)
```

```python
dt.fit(x_train,y_train)
```

```python
predictions1 = dt.predict(x_test)

from sklearn.metrics import accuracy_score
accuracy_score(y_test,predictions1)
```

```python
#Random Forest
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100,max_depth=5,random_state=1)
```

```python
rf.fit(x_train,y_train)
```

```python
predictions2 = rf.predict(x_test)
accuracy_score(y_test,predictions2)
```

```python
x_train,x_validation,y_train,y_validation = train_test_split(x_train,y_train,test_size=0.3)
```

```python
#Neural Network
import tensorflow as tf

model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(units=32,input_shape=(30,)))
model.add(tf.keras.layers.Dense(units=30,activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(units=30,activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(units=1,activation="sigmoid"))

model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["acc"])
model.fit(x_train,y_train,epochs=10,validation_data=(x_validation,y_validation))
```

```python
predictionsnn = model.predict(tst)
predictionsnn = (predictionsnn > 0.5).astype(int).ravel()
```
