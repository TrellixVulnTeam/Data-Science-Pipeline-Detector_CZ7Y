```python
import numpy as np
import pandas as pd
import pickle
pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

```python
comp_name = "tmay"
```

```python
def check_number(element):
    element = str(element)
    if "." in element:
        partition = element.partition('.')
        if (partition[0].isdigit() and partition[1] == '.' and partition[2].isdigit()) or (partition[0] == '' and partition[1] == '.' and partition[2].isdigit()) or (partition[0].isdigit() and partition[1] == '.' and partition[2] == ''):
            return True 
        return False
    else:
        if element.isdigit():
            return True 
        return False

    
with open(f"../input/configs-{comp_name}/Table.pkl", "rb") as f:
    Table = pickle.load(f)
Table.drop(["_dataset","aug_type", "use_cutmix"], axis=1)

# highlight = lambda x: ['background: yellow' if x.bv ==  Table.bv.max() else "" for i in x]
# Table.drop(["_dataset","aug_type", "use_cutmix"], axis=1).style.apply(highlight, axis=1)
```

```python
exp_no = Table[Table.bv== Table.bv.max()].exp_no.values[0]
with open(f"../input/configs-{comp_name}/log_exp_{exp_no}.pkl", "rb") as f:
    log_table_sample = pickle.load(f)
highlight = lambda x: ['background: yellow' if x.trial_score ==  log_table_sample.trial_score.max() else "" for i in x]
print("Exp with best score so far: ",exp_no)
display(log_table_sample.style.apply(highlight, axis=1))
```

```python
for e in Table.exp_no.values: #range(9): #Table.exp_no.values:
    exp_no = e 
    with open(f"../input/configs-{comp_name}/log_exp_{exp_no}.pkl", "rb") as f:
        log_table_sample = pickle.load(f)
    print("Experiment no: ",exp_no)
    
#-> for displaying log table
#     highlight = lambda x: ['background: yellow' if x.trial_score ==  log_table_sample.trial_score.max() else "" for i in x]
#     display(log_table_sample.style.apply(highlight, axis=1))

    print()
    
    with open(f"../input/configs-{comp_name}/log_exp_{exp_no}.pkl", "rb") as f:
        log_table_sample = pickle.load(f)
    log_table = log_table_sample

    import keras
    from matplotlib import pyplot as plt

    color_list = ["red","pink","green", "blue", "cyan","magenta","black","orange","violet","indigo"]
    color_list += ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']
    
    #=====================================================================================
    counter = 0 
    plt.figure(figsize=(20,8))
    for ind,i in  enumerate(log_table.columns):
        if sum([not check_number(j) for j in log_table[i].values]) == 0:
            if i == "trial_score":
                plt.plot(log_table["trial_score"].values**20,  label = i,color=color_list[counter])
                v = [Table[Table.exp_no == exp_no].bv[exp_no]**20 for _ in range(len(log_table[i]))]
                plt.plot(v, "-.", color= color_list[counter])
            else:
                plt.plot(log_table[i].values,  label = i,color=color_list[counter])
                v = [Table[Table.exp_no == exp_no].bp[exp_no][i] for _ in range(len(log_table[i]))]
                plt.plot(v, "-.", color=color_list[counter])
            counter += 1
    plt.axvline(x = log_table[log_table.trial_score== Table[Table.exp_no == exp_no].bv[exp_no]].index[0], color = 'black', label = 'best trial')

    plt.plot()
    plt.title('model auc')
    plt.ylabel('auc')
    plt.xlabel('trial')
    plt.legend(loc='upper left')

    plt.legend(loc=(1.02, 0.6))
    plt.yscale("log")
    #plt.xscale("log")

    plt.show()
    #=====================================================================================
    
    #=====================================================================================
    counter = 0 
    i = "trial_score"
    plt.figure(figsize=(20,5)) #(10,5))
    plt.plot(log_table["trial_score"].values.clip(Table[Table.exp_no == exp_no].bv[exp_no]-10**-3,Table[Table.exp_no == exp_no].bv[exp_no]+10**-3), marker="o",ls="--",markersize=3,markeredgewidth=6, markeredgecolor="y", markerfacecolor="b",   label = i,color=color_list[counter])
    v = [Table[Table.exp_no == exp_no].bv[exp_no] for _ in range(len(log_table[i]))]
    plt.plot(v, "-.", color= color_list[counter],label="best_score")
    plt.axvline(x = log_table[log_table.trial_score== Table[Table.exp_no == exp_no].bv[exp_no]].index[0], color = 'black', label = 'best trial')

    #plt.yscale("log")
    plt.title('model auc')
    plt.ylabel('auc')
    plt.xlabel('trial')
    plt.legend(loc='upper left')

    plt.legend(loc=(1.02, 0.6))

    plt.show()
    #=====================================================================================
    
    print("="*100)
    
```
