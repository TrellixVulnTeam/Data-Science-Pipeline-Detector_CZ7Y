```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
train['f_27']=le.fit_transform(train['f_27'])
test['f_27']=le.fit_transform(test['f_27'])
```

```python
x_train=train.drop(['id','target','f_20','f_19','f_25','f_16'],axis=1)
x_test=test.drop(['id','f_20','f_19','f_25','f_16'],axis=1)
```

```python
y_train=train['target']
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rd=RandomForestClassifier(criterion='gini',max_depth=5,n_estimators=500)
```

```python
rd.fit(x_train,y_train)
```

```python
y_pred=rd.predict(x_test)
```
