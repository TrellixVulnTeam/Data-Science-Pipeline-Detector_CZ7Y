Pandas Profiling generates report from a Dataframe. We already know some methods like df.describe() or df.info() that helps us a lot but with a simple command we can save time in our explorations having a wide idea about our data.

You can have informations about:
- Type of columns
- Unique values
- Missing values
- descriptive statitics( mean, median standard deviation and so on)
- Most frequent values
- Histograms
- Correlations
- Duplicate rows
- Text analysis
- Image and file analysis


You can have more info about Pandas profiling on the following link:

https://github.com/ydataai/pandas-profiling

```python
from pandas_profiling import ProfileReport 
```

In this case we have almost every variable who is numeric as we can control with this command:

```python
train=pd.read_csv("/kaggle/input/titanic/train.csv")
train.head()
```

```python
#The only object variable is f_27
train.info()
```

```python
#We can have a first glance about the distributions withthe df.describe() method as we see below:
train.describe()
```

```python
profile_train = ProfileReport(train, title="Train Profiling Report")
```

```python
profile = ProfileReport(train, title="Train Profiling Report")
```

```python
ProfileReport(train)
```

If you want to generate an HTML Report File you can use this function.

```python
profile.to_file("your_report.html")
```
