```python
train_df = pd.read_csv("/kaggle/input/tabular-playground-series-may-2022/train.csv")
test_df = pd.read_csv("/kaggle/input/tabular-playground-series-may-2022/test.csv")
```

```python
print("Size of the train_df :",train_df.shape[0])
print("Size of the test_df: ",test_df.shape[0])
```

```python
## check null values
train_df.isnull().sum()
```

```python
print("Total columns are ",train_df.shape[1])
num_cols = train_df.select_dtypes(["int","float"])
print("Number of numerical columns are : ",num_cols.shape[1])
cat_cols = train_df.select_dtypes(["object"])
print("Number of categorical columns are : ",cat_cols.shape[1])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

target = train_df.target
sns.countplot(x=target.value_counts())
plt.show() ## data seems balanced
```

```python
## creating a model with out modification how do it perform ?
## we can get idea of what to do since it is binary classification problem we could use logistic regression 
```

```python
from sklearn.model_selection  import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
train_x, cv_x, y_train, y_cv  = train_test_split(num_cols,train_df.target,stratify=train_df.target)
```

```python
train_x.drop(['target',"id"],axis=1,inplace=True)
```

```python
cv_x.drop(["target","id"],axis=1,inplace=True)
```

```python
from sklearn.metrics import accuracy_score
```

```python
lr = LogisticRegression(max_iter=500)
lr.fit(train_x.values, y_train.values)
pred = lr.predict(train_x.values)
print("The train accuracy : ",accuracy_score(y_train.values,pred))
pred  = lr.predict(cv_x.values)
print("The cv accuracy : ",accuracy_score(y_cv.values, pred))
```

```python
## submission
test_x = test_df.select_dtypes(["int","float"])
test_x.shape
test_id = test_x.id.values
test_x.drop("id",axis=1,inplace=True)
```

```python
pred = lr.predict(test_x.values)
```

```python
submission = pd.read_csv("/kaggle/input/tabular-playground-series-may-2022/sample_submission.csv")
submission.head()
```

```python
submission_df = pd.DataFrame({
    "id" : test_id,
    "target": pred
})
```

```python
submission_df.head()
```

```python
submission_df.to_csv("submission.csv",index=False)
```
