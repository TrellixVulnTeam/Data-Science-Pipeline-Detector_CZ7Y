# **Import necessary Library**

```python
import numpy as np
import pandas as pd

import torch
import flash
from flash.tabular import TabularClassificationData, TabularClassifier
```

# **Feature selection, Model initialization, Predict the output**

```python
datamodule = TabularClassificationData.from_csv(
    categorical_fields=['f_27'],
    numerical_fields=['f_00'],
    target_fields="target",
    train_file="../input/tabular-playground-series-may-2022/train.csv",
    val_split=0.1,
    batch_size=8,
)
```

```python
# 2. Build the task
model = TabularClassifier.from_data(datamodule, backbone="fttransformer")
```

```python
# 3. Create the trainer and train the model
trainer = flash.Trainer(max_epochs=1, gpus=torch.cuda.device_count())
trainer.fit(model, datamodule=datamodule)
```

# **Save the model**

```python
# 5. Save the model!
trainer.save_checkpoint("tabular_classification_model.pt")
```
