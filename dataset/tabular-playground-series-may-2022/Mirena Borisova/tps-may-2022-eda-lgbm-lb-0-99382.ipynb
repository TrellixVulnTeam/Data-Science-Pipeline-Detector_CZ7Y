```python
def val_count_df(df, column_name, sort=True):
    value_count = df[column_name] \
        .value_counts(sort=sort) \
        .reset_index() \
        .rename(columns={column_name: 'Value Count',
                         'index': column_name})
    
    value_count['Percentage'] = df[column_name].value_counts(sort=sort, normalize=True) * 100
    value_count = value_count.reset_index()
    
    return value_count
```

```python
def feature_engineering(df):
    new_df = df.copy()
    
    for letter in ['A', 'B', 'C', 'D', 'E']:
        new_df[letter + '_count'] = new_df['f_27'].str.count(letter)
        
    for i in range(10):
        new_df[f'f_27_{str(i)}'] = new_df['f_27'].str[i].apply(lambda x: ord(x) - ord('A'))
        
    new_df['unique_characters'] = new_df['f_27'].apply(lambda x: len(set(x)))
    
    new_df = new_df.drop(columns=['f_27', 'id'])
    
    return new_df
```

```python
%%time

train_df = feature_engineering(train_df)
test_df = feature_engineering(test_df)
```

```python
from lightgbm import LGBMClassifier
from sklearn.metrics import roc_auc_score, accuracy_score
from sklearn.model_selection import StratifiedKFold
```

```python
y = train_df['target']
X = train_df.drop(columns=['target'])
X_test = test_df
X.head(2)
```

```python
model = LGBMClassifier(n_estimators=10000, 
                       learning_rate=0.1, 
                       random_state=0, 
                       min_child_samples=90,
                       num_leaves=150,
                       max_bins=511,
                       n_jobs=-1)
```

```python
def k_fold_cv(model, X, y):
    kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)
    
    feature_imp, y_pred_list, y_true_list, roc_list, acc_list = [], [], [], [], []
    for fold, (train_idx, valid_idx) in enumerate(kfold.split(X, y)):
        if fold < 2:
            print(f'FOLD: {fold}')
            X_train = X.loc[train_idx]
            X_valid = X.loc[valid_idx]
            
            y_train = y.loc[train_idx]
            y_valid = y.loc[valid_idx]
            
            model.fit(X_train, y_train)
            
            y_pred = model.predict_proba(X_valid)[:, 1]
            
            y_pred_list = np.append(y_pred_list, y_pred)
            y_true_list = np.append(y_true_list, y_valid)
            
            roc_list.append(roc_auc_score(y_valid, y_pred))
            acc_list.append(accuracy_score(y_pred.round(), y_valid))
            
            print(f'ROC_AUC_SCORE: {roc_auc_score(y_valid, y_pred)}')
            print(f'ACCURACY_SCORE: {accuracy_score(y_pred.round(), y_valid)}')
            
            try:
                feature_imp.append(model.feature_importances_)
            except AttributeError:
                    pass
                
    return feature_imp, y_pred_list, y_true_list, roc_list, acc_list, X_valid, y_valid
```

```python
%%time

feature_imp, y_pred_list, y_true_list, roc_list, acc_list, X_valid, y_valid = k_fold_cv(model=model, 
                                                                                        X=X, 
                                                                                        y=y)
```

```python
print(f'ACUURACY_SCORE.MEAN(): {np.mean(acc_list)}')
print(f'ROC_AUC_SCORE.MEAN(): {np.mean(roc_list)}')
```

```python
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix

def plot_conf_matrix(preds, true, ax=None):
    cm = confusion_matrix(preds.round(), true)
    cmd = ConfusionMatrixDisplay(confusion_matrix=cm)
    cmd.plot(ax=ax, colorbar=False, values_format='.6g')
    plt.grid(False)
    
    return cmd
```

```python
def fold_feature_importances(model_importances, 
                             column_names, 
                             model_name, 
                             n_folds=5, 
                             ax=None, 
                             boxplot=False):
    importances_df = pd.DataFrame({'feature_cols': column_names, 
                                   'importances_fold_0': model_importances[0]})
    
    for i in range(1, n_folds):
        importances_df[f'importances_fold_{str(i)}'] = model_importances[i]
        
    importances_df['importances_fold_median'] = importances_df.drop(columns=['feature_cols']).median(axis=1)
    importances_df = importances_df.sort_values(by='importances_fold_median', ascending=False)
    
    if ax == None:
        fig, ax = plt.subplots(figsize=(15, 25))
    
    if boxplot == False:
        ax = sns.barplot(data=importances_df, 
                         x='importances_fold_median', 
                         y='feature_cols', 
                         palette='spring')
        ax.set_xlabel('MEDIAN FEATURE IMPORTANCE ACROSS ALL FOLDS')
    elif boxplot == True:
        importances_df = importances_df.drop(columns='importances_fold_median')
        importances_df = importances_df \
            .set_index('feature_cols') \
            .stack() \
            .reset_index() \
            .rename(columns={0: 'feature_importance'})
        
        ax = sns.boxplot(data=importances_df, 
                         y='feature_cols', 
                         x='feature_importance', 
                         palette='turbo',
                         orient='h')
        ax.set_xlabel('FEATURE IMPORTANCE ACROSS ALL FOLDS')
        
    plt.title(model_name)
    ax.set_ylabel('FEATURE COLUMNS')
    
    return ax
```

```python
f, ax = plt.subplots(figsize=(15, 20))

fold_feature_importances(model_importances=feature_imp,
                         column_names=X_valid.columns,
                         model_name="LGBM",
                         n_folds=2,
                         ax=ax,
                         boxplot=False)
```
