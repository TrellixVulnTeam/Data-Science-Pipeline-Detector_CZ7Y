```python
import pandas as pd
import scipy
from scipy import stats

sub = pd.read_csv('../input/tabular-playground-series-may-2022/sample_submission.csv')
sub_823 = pd.read_csv('../input/lightautoml-0-99807/nn_preds_only.csv')
sub_822 = pd.read_csv('../input/auto-blending-0-99822/auto_blending_ensemble.csv')
sub_821 = pd.read_csv('../input/auto-blending-0-99821/auto_blending.csv')
sub_820 = pd.read_csv('../input/advanced-keras-0-99820/advanced_keras.csv')
```

```python
sub_823['target'] = scipy.stats.rankdata(sub_823['target'])
sub_822['target'] = scipy.stats.rankdata(sub_822['target'])
sub_821['target'] = scipy.stats.rankdata(sub_821['target'])
sub_820['target'] = scipy.stats.rankdata(sub_820['target'])

sub['target'] = sub_823['target'] * 0.3 + \
                sub_822['target'] * 0.3 + \
                sub_821['target'] * 0.3 + \
                sub_820['target'] * 0.3

sub.to_csv('ensemble.csv', index=False)
sub
```
