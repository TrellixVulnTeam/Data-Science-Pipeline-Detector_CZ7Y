```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
```

```python
train = pd.read_csv('../input/predict-who-is-more-influential-in-a-social-network/train.csv')
test = pd.read_csv('../input/predict-who-is-more-influential-in-a-social-network/test.csv')
```

```python
print(train.shape,test.shape)
```

```python
#Checking for null values
print('Null Values in Train:',train.isna().values.any())
print('Null Values in Test:',test.isna().values.any())
```

```python
cor_matrix = train.corr()
```

```python
#Upper Triangle Matrix
upper_tri = cor_matrix.where(np.triu(np.ones(cor_matrix.shape),k=1).astype(np.bool))
```

```python
#Columns having high covarience
drop = [column for column in upper_tri.columns if any(upper_tri[column] > 0.95)]
print(drop)
```

```python
#Selecting features based on cor_matrix result
features=['A_follower_count','A_listed_count','A_mentions_received','A_posts','A_network_feature_1','A_network_feature_2','A_network_feature_3','B_follower_count','B_following_count','B_listed_count','B_mentions_received','B_posts','B_network_feature_1','B_network_feature_2','B_network_feature_3']
target = 'Choice'
```

```python
X = train[features]
y = train[target]
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200,random_state= 24)
```

```python
rf.fit(X,y)
```

```python
y_hat = rf.predict(test[features])
```

```python
sub = pd.DataFrame({"Id": list(range(1,len(y_hat)+1)),
                         "Choice":y_hat})
```

```python
sub.to_csv('submission.csv',index=False)
```
