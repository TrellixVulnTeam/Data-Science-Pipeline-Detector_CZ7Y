> # Kaggle's third annual Machine Learning and Data Science Survey

## Executive Summary

Kaggle's third annual Machine Learning and Data Science Survey is an industry-wide survey that presents a truly comprehensive view of the state of data science and machine learning. This report aim to tell the complete story for kaggle participants who's their ages are from 18-34 or Millennials 

## Challenge Overview

Out of 19,717 usable respondents from 171 countries, this exclusive summary will focus on **(69.43%) who's their ages are from 18-34 years** or Millennial. <br/><br/>

The analysis of this report will based on EDA to explore the main characteristics of Millennial in kaggle community such as **demographic status**, **Working environment** , we will also put highlight on their **learning path**, **technology adapted** their **weakness or skill shortage** and many other things. we also will identify any differences for these characteristic based on gender and continent.<br/><br/>

In general we see that kaggle community of millennial are mostly male and have university degree, they having little experience in data science fields and consider kaggle their favorite platforms for learning data science, you will see that North America has the highest pay and most spending on machine learning and cloud computing products.



## Report Methodology

Kaggle’s State of Machine Learning and Data Science 2019 survey has about 19,717 respondents from 171 countries a round the world, the country which has less than 50 respondents was grouped to "Others" which later named as "Not classified". The respondent were found through kaggle channels, email list, discussion forums and social media channels from October 8th to October 28th in 2019, However this reports only focused on millennial respondent whose their ages are from 18-34 years.<br/><br/>

Multiple choice multiple response questions were split into multiple columns to give the respondent the ability to select more than one option, so for many charts where the total percentage is above 100%
All monetary amounts captured in the report are in USD.



## Main Findings<br/>
* Most of Millennial in kaggle community are Males.
* Half of Millennial are coming from Asia.
* More than (80%) of Millennial has university degree and more than (50%) has advanced university degree.
* Most of Millennial work at either small  or very large organizations.
* Half of Millennial have less than 2 years of experience in writing codes.
* Millennial in North America has the highest compensation and Asia has the lowest.
* North America is most continent spending on machine learning and cloud computing products.
* Jupyter is most common analytics tool among all other IDE for kaggle community of millennial.
* Python is most common programming language for data science for kaggle community of millennial.
* Kaggle is millennial favorites media for data science and coursera is the best teaching platform.
* Millennial are most Appling simple ML method such as linear and logistic regression, Decision trees and random forest in their regular work using sicit_learn, their favorite ML package.
* Millennial are not familiar with cloud computing, big data/analytics and machine learning products.





```python

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

import pandas as pd
from pandas import Series,DataFrame
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
file_path="/kaggle/input/kaggle-survey-2019/multiple_choice_responses.csv"
kaggle_O=pd.read_csv(file_path)
kaggle_O=kaggle_O[1:]
```

```python
kaggle_O.Q1=kaggle_O.Q1.astype("category")
kaggle_O["Age_generation"]=kaggle_O.Q1
kaggle_O["Age_generation"]=kaggle_O["Age_generation"].replace(["35-39","40-44","45-49"],"Adults")
kaggle_O["Age_generation"]=kaggle_O["Age_generation"].replace(["18-21","22-24","25-29","30-34"],"Millennials")
kaggle_O["Age_generation"]=kaggle_O["Age_generation"].replace(["50-54","55-59","60-69","70+"],"elderly")

kaggle_O.Q3=kaggle_O.Q3.astype("category")
kaggle_O["continent"] =kaggle_O["Q3"]
kaggle_O["continent"]= kaggle_O["continent"].replace(["Malaysia","Philippines","Saudi Arabia",'Iran, Islamic Republic of...',"Viet Nam",'Republic of Korea',"India", 'Russia',"Greece","Pakistan","Japan","South Korea","Indonesia",'Hong Kong (S.A.R.)',"Turkey","Singapore","Israel","Taiwan",'Bangladesh','Thailand', 'China'],"Asia")
kaggle_O["continent"]= kaggle_O["continent"].replace(["Algeria",'Tunisia',"Nigeria","Morocco","South Africa","Egypt","Kenya"],"Africa")
kaggle_O["continent"]= kaggle_O["continent"].replace(['Austria',"Belgium","Romania",'Denmark','France','Germany','Netherlands','Italy','Ireland',"Ukraine",'Portugal',"Switzerland","Chile","Hungary","Norway","Belarus","Sweden","United Kingdom of Great Britain and Northern Ireland","Poland","Spain","Czech Republic"],"Europe")
kaggle_O["continent"]= kaggle_O["continent"].replace(["United States of America","Mexico","Canada"],"North America")
kaggle_O["continent"]= kaggle_O["continent"].replace(["Brazil","Argentina","Colombia",'Peru'],"South America")
kaggle_O["continent"]= kaggle_O["continent"].replace(['Australia',"New Zealand"],"'Australia")
kaggle_O["continent"]= kaggle_O["continent"].replace(["Other"],"Not classified")

kaggle_O.Q5=kaggle_O.Q5.astype("category")
kaggle_O["Employed"]=kaggle_O["Q5"]
kaggle_O["Employed"]=kaggle_O["Employed"].replace(["Data Scientist","Software Engineer","Other","Data Analyst","Research Scientist",
                                               "Business Analyst","Product/Project Manager","Data Engineer","Statistician",
                                               "DBA/Database Engineer"],"Employed")

kaggle_O.Q4=kaggle_O.Q4.astype("category")
kaggle_O["Educated"]=kaggle_O["Q4"]
kaggle_O["Educated"]=kaggle_O["Educated"].replace(["Master’s degree","Bachelor’s degree","Doctoral degree","Professional degree "],"Educated")
```

```python
kaggle=kaggle_O[(kaggle_O.Age_generation=="Millennials")]
```
