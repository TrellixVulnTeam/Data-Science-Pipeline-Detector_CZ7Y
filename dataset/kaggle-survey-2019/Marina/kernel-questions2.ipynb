```python
def autolabel(rects, labels=None, height_factor=1.01):
    for i, rect in enumerate(rects):
        height = rect.get_height()
        if labels is not None:
            try:
                label = labels[i]
            except (TypeError, KeyError):
                label = ' '
        else:
            label = '%d' % int(height)
        ax.text(rect.get_x() + rect.get_width()/2., height_factor*height,
                '{}'.format(label),
                ha='center', va='bottom')
#autolabel(ax.patches, labels, height_factor=1.01)
#если не указывать labels при вызове функции, то напечатаются величины
#ax.patches - рисунок.patches
```

```python
#my_df загруженный csv файл
#dict_slice список значений индексов столбцов разбитого на несколько частей вопроса
#questions_text текст вопроса- имя графика
def newfunc(my_df, dict_slice, questions_text, other_text = None):
    df = my_df.iloc [:, dict_slice]
    name = [i for i in df.columns]
    data = []
    columns = name
    for i in df.columns:
        data.append(df[i].isnull().sum())
    data = np.array(data)
    df1 = pd.DataFrame(columns=columns)
    df1.loc[len(df1)] = data
    df1 = df1.T.reset_index()
    df1.columns = ['questions', 'count']
    plt.figure(figsize=(8,6))
    plt.xticks(rotation=45, horizontalalignment='right')
    if other_text != None:plt.title(other_text+'  '+questions_text)
    else: plt.title(questions_text)
    ax = sns.barplot(x='count', y='questions', data=df1);
    
    # set individual bar lables using above list
    for i in ax.patches:
    # get_width pulls left or right; get_y pushes up or down
        ax.text(i.get_width()+.3, i.get_y()+.38, str(int(i.get_width()/len(my_df)*100))+'%', fontsize=15,color='black')
    
    plt.show()
```

```python
 def to_frequency_table(data): 
    frequencytable = {} 
    for key in data: 
        if key in frequencytable: frequencytable[key] += 1 
        else: frequencytable[key] = 1 
    return frequencytable 
```

```python
def new_df(df, i):
    df = pd.DataFrame([to_frequency_table(df[i])])
    df = df.T.reset_index()
    df.columns = ['answer', 'count']
    g = sns.catplot(y = 'count', x="answer",kind="bar",  data=df);
    g.fig.set_figheight (6)
    g.fig.set_figwidth (70)
    for ax in g.axes.flat:
        ax.set_xticklabels(ax.get_xticklabels(), rotation=45, size=15, horizontalalignment='right')
        ax.set_yticklabels(ax.get_yticklabels(), size='large')   
    
    plt.title(i)    
```

```python
def new_df_drop(df, i):
    df = pd.DataFrame([to_frequency_table(df[i])])
    df = df.T.reset_index()
    df.columns = ['answer', 'count']
    try:
        df = df.loc[df['count'] > 1]
        g = sns.catplot(y = 'count', x="answer",kind="bar",  data=df);
        g.fig.set_figheight (6)
        g.fig.set_figwidth (30)
        for ax in g.axes.flat:
            ax.set_xticklabels(ax.get_xticklabels(), rotation=45, size=10, horizontalalignment='right')
            ax.set_yticklabels(ax.get_yticklabels(), size='large')
        
        for p in ax.patches:
            ax.text(p.get_x() + 0.1, p.get_height() * 1.1, int(p.get_height()), color='k',\
                    size='medium')
        
       
        plt.title(i, size='large')
    except ValueError:
        print(i)
        print('No responses with this question')
```
