```python
def age_categorization(x):
    if (x == '18-21'):
        return 'young'
    if (x == '22-24') | (x == '25-29')|(x == '35-39')| (x =='40-44'):
        return 'mid_age'
    return 'above_mid_age'
mul_wo_question['Q1'] = mul_wo_question.Q1.apply(age_categorization)
```
