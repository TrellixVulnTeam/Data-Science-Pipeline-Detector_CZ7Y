```python
def numriser(v):
    v1=[]
    for i in v:
        v1=np.append(v1,int(i))
    return(v1)
```

```python
def co(c):
    t1=[]
    for i in np.arange(len(t)):
        if t[i]>c:
            t1=np.append(t1,t[i])   
    return(t1)


def dnumriser(c):
    v2=[]
    for i in np.arange(1,len(c)):
        if len(c[i])==5:
            v1=(int(c[i][0:2])+int(c[i][3:5]))/2
            v2=np.append(v2,v1)
        else:
            v1=(int(c[i][0:2]))
            v2=np.append(v2,v1)
    return (v2)

def cc(v,n):
    f=0
    for i in np.arange(1,len(v)):
        if (v[i])==n:
            f=f+1
    return (f)
```
