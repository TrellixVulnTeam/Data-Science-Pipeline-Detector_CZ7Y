```python
# for cleaning and extracting the questions
def get_ques_and_clean_df(df):
    cols = df.columns
    names = df.iloc[0,:]
    questions = {a : b  for a, b in zip(cols , names)}
    df.drop([0] , inplace = True)
    df.reset_index(inplace=True)
    df.drop(labels='index',axis=1, inplace=True)
    return questions

# for printing the questions
def print_questions(questions):
    print(f'total number of questions  = {len(questions)}')
    for i in questions.keys():
        print(f'{i} : {questions[i]} \n')
```

```python
# function to get the top5 df
def _get_top5_df(df , label):
    df[label] = df[label].replace('United States of America' , 'USA')
    df[label] = df[label].replace('United States' , 'USA')
    df[label] = df[label].replace("People 's Republic of China" , 'China')
    df[label] = df[label].replace('United Kingdom' , 'UK')
    top_countries_value = df[label].value_counts()
    others = top_countries_value.iloc[7:].sum()
    top_countries_value = top_countries_value.iloc[0:6]
    top_countries_value.append(pd.Series(data=[others] , index=['others']))
    top_countries_percent = top_countries_value.copy()
    for  i in top_countries_percent:
        top_countries_percent =  top_countries_percent.replace(i , (i/ top_countries_value.sum()) * 100)
    return top_countries_value , top_countries_percent
```

```python
#function to extract yearwise features for male and female
def obtain_yearswise_features(label1 : list , label2: list):
    years = [2017 , 2018 , 2019 , 2020]
    _2017 = India_2017.loc[India_2017[label1[0]] == 'Male' , label2[0]].value_counts()
    _2018 = India_2018.loc[India_2018[label1[1]] == 'Male' ,label2[1]].value_counts()
    _2019 = India_2019.loc[India_2019[label1[2]] == 'Male' ,label2[2]].value_counts()
    _2020 = India_2020.loc[India_2020[label1[3]] == 'Male' ,label2[3]].value_counts()

    yearwise_male = pd.concat([_2017 , _2018 , _2019 , _2020] , axis=1)
    yearwise_male.columns = years
    
    _2017 = India_2017.loc[India_2017[label1[0]] == 'Female' , label2[0]].value_counts()
    _2018 = India_2018.loc[India_2018[label1[1]] == 'Female' ,label2[1]].value_counts()
    _2019 = India_2019.loc[India_2019[label1[2]] == 'Female' ,label2[2]].value_counts()
    _2020 = India_2020.loc[India_2020[label1[3]] == 'Female' ,label2[3]].value_counts()

    yearwise_female = pd.concat([_2017 , _2018 , _2019 , _2020] , axis=1)
    yearwise_female.columns = years
    return yearwise_male , yearwise_female
```

```python
# for converting numbers in string form to proper machine understood numerals
def decoma(x):
    if type(x) == str:
        symbols = [',' ,'$' , '<' , ">" , '+']
        for sym  in symbols:
            x = x.replace(sym , '')
        if len(x) <= 3:
            x += '000'
    return x
```

```python
# to genrate the bin labels and bins in proper sorted order
def generate_bins(df , label):
    # preparing bins and labels for binning
    sal_bins = df[label].unique()
    labels_sal = sal_bins.copy()
    labels_sal = list(labels_sal)
    labels_sal.pop(labels_sal.index(np.nan))
    sal_bins = [a.split('-')[0] for a in sal_bins if type(a) == str ]
    sal_bins = [int(decoma(x)) for x in sal_bins]
    sal_bins.sort()
    sal_bins.append(np.NaN)
    # list comprehension to sort the string intervals in correct order using age_bins list
    bin_labels_sal = [c for a in sal_bins for c in labels_sal if int(decoma(c.split('-')[0])) == a]
    return sal_bins , bin_labels_sal
```

```python
def plot_sal(df , label1 ,label2 , fig , row  , showlegend=False):
    #extracting data for common 
    sal_df = df[[label1 , label2]]
    #genrating bin labels
    _ , labels = generate_bins(sal_df, label2)
    #getting total salary distribtuion
    df_main = pd.DataFrame(sal_df[label2].value_counts())
    df_main = df_main.reindex(labels)
    # drawing the first graph row =x , col = 2*x - 1
    fig.add_trace(go.Funnel(x=df_main.index , y=df_main[label2] , orientation='v',width=0.8, text=df_main[label2], showlegend=False), row=row , col=1)
    fig.update_traces(texttemplate='%{text:.2s}', textposition='outside' , textfont= dict(size=14 , family='Aerial, sans-serif'), textangle=0, row = row , col=1)
    #plotting gender wise
    color = {'Male':'green' , 'Female':'yellow'}
    for g in sal_df[label1].unique():
        df_sub = sal_df.loc[sal_df[label1] == g]
        df_sub = pd.DataFrame(df_sub[label2].value_counts())
        df_sub = df_sub.reindex(labels)
        fig.add_trace(go.Bar(y=df_sub.index , x = df_sub[label2] , orientation='h' ,name=g, marker=dict(color=color[g]) , showlegend=showlegend), row=row , col=2)
        
```

```python
fig = make_subplots(rows=3 , cols=2,
                    horizontal_spacing= 0.14, column_widths= [0.55 , 0.45],
                   subplot_titles = ('2018' , '2018 (by gender)',
                                    '2019' , '2019 (by gender)',
                                    '2020' , '2020 (by gender)'))
India_2018 = India_2018.loc[India_2018['Q9'] != 'I do not wish to disclose my approximate yearly compensation']
plot_sal(India_2018 , 'Q1' , 'Q9' ,fig , 1, showlegend=True)
plot_sal(India_2019 , 'Q2' , 'Q10' , fig , 2)
plot_sal(India_2020 , 'Q2' , 'Q24' , fig , 3)
fig.update_layout(height=1300, width=1000 , template='seaborn' , 
                  legend = dict(orientation = 'h' ,
                               yanchor = 'bottom',
                               y =1.02,
                               xanchor ='right',
                               x=1),
                 margin = dict(t=150,
                               l=30,
                               r = 30,
                               b= 40),
                 title_text = 'FIGURE 8 : <br>Distribution of yearly compensation(salary) for ML and DS in India accross genders',
                 title = dict(xanchor='left' , x=0.02))
fig.show()
```

```python
def get_counts(df , labels :list , name):
    d_f = df[labels[0]].value_counts()
    for f in labels[1:]:
        d_f = d_f.append(df[f].value_counts())
    d_f = pd.DataFrame(d_f, columns=[name])
    return d_f
```

```python
def clean_df(df):
    df.fillna(0 , inplace=True)
    years = [2019 , 2020]
    for year in years:
        df.loc['C/C++' , year] = df.loc['C' , year] + df.loc['C++' , year]
    
    df.drop(['C' , 'C++'], axis='index', inplace=True)
    for year in years:
        df.loc['Javascript/Typescript' , year] = df.loc['Javascript' , year] + df.loc['TypeScript' , year]
    df.drop(['TypeScript' , 'Javascript'] , axis='index' ,inplace=True)
    df.drop(['Other'] , inplace=True)
```

```python
def generate_dfs(dfs : list , labels : list , names : list):
    d_f = get_counts(dfs[0] , labels[0] , names[0])
    for i , df in enumerate(dfs[1 :]):
        d_f = pd.concat([d_f , get_counts(df , labels[i+1] , names[i+1])] , axis=1)
    clean_df(d_f)
    return d_f
        
```
