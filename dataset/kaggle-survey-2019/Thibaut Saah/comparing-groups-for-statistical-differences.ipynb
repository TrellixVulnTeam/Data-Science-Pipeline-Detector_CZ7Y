# Education and country

```python
contengency = pd.crosstab(multiple_choice_responses.Q3.drop([0]), multiple_choice_responses.Q4.drop([0]))
contengency.sort_values(by = ["Bachelorâ€™s degree"], ascending=False).head(10)
```

```python
import scipy.stats as st
st_chi2, st_p, st_dof, st_exp = st.chi2_contingency(contengency)
print('Khi2_calculate:', st_chi2, 'P value of test:', st_p, 'degree of freedom:', st_dof)
```

The Chi-Square test of independence is used to determine if there is a significant relationship between country and high level school variables if we are a datascientist. 
The distribution of datascientist per country and according to their high level school is not identical.

```python
heatData1 = (contengency.values - st_exp)/st_exp
heatData1 = pd.DataFrame(heatData1, columns=contengency.columns, index = contengency.index)
plt.figure(figsize=(20, 20))
sns.heatmap(heatData1, cmap="YlGnBu", annot=True)
```

There is a lot information into the above picture. In Chile for example there are 5.8 times more datascientist with professional degree in chile than what is expected.  ie if we expected 1000 datascientist with professional degree in Chile according to average of datascientist in the world with professional, there are 5800 datascientist instead of 1000. 

```python
B = heatData1[(heatData1["Doctoral degree"]<-0.5) | (heatData1["Doctoral degree"]>0.5)].sort_values(by = ["Doctoral degree"])
height = B["Doctoral degree"].values*100
var = B.index.values
plt.figure(figsize = (15,10))
pos = np.arange(len(var))
bars = plt.barh(pos, height)
plt.yticks(pos, var)
plt.xlabel("deviation from statistical independence")
plt.title('statistical difference (Doctoral degree as high level school versus country)')
plt.tick_params(top = False, left = False, bottom = False, labelbottom = False)
# remove frames
for spine in plt.gca().spines.values():
    spine.set_visible(False)

for bar in bars:
    plt.gca().text(bar.get_width()+1.2, bar.get_y()+bar.get_height()/2, str(int(bar.get_width()))+"%")
    
```

As you can see according to the above barplot, in switzerland there are almost 3 times more datascientist than is expected

```python
heatData1.columns
```

```python
C = heatData1[(heatData1["No formal education past high school"]<=-1) | (heatData1["No formal education past high school"]>1)].sort_values(by = ["No formal education past high school"])
height = C["No formal education past high school"].values*100
var = C.index.values
plt.figure(figsize = (15,10))
pos = np.arange(len(var))
bars = plt.barh(pos, height)
plt.yticks(pos, var)
plt.xlabel("deviation from statistical independence")
plt.title('No formal education past high school')
plt.tick_params(top = False, left = False, bottom = False, labelbottom = False)
# remove frames
for spine in plt.gca().spines.values():
    spine.set_visible(False)

for bar in bars:
    plt.gca().text(bar.get_width()+1.2, bar.get_y()+bar.get_height()/2, str(int(bar.get_width()))+"%")
    
```

Interpreting is not difficult, a 337% for Roumania simply means that the number of datacientist present in this state are 4,34 times higher than what's expected.
In Morocco it 's 2 times slower than what's expected.
There are some insights. 

Another important relevant information is about those people with master degree as high level school in each state working in the field of data. they are substantially equal in number from a statistical point of view.
