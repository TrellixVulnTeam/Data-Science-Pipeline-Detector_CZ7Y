# Data Scientists in 2019 - Kaggle Survey


# Introduction
We will analyze the dataset 2018 Kaggle ML & DS Survey Challenge with answers provided by the respondents to the survey of Kaggle users in 2019.
This survey received 19,717 usable respondents from 171 countries and territories. If a country or territory received less than 50 respondents, we grouped them into a group named “Other” for anonymity.

Survey Detail - This year, as in 2017 and 2018, we set out to conduct an industry-wide survey that presents a truly comprehensive view of the state of data science and machine learning. The survey was live for three weeks in October, and after cleaning the data we finished with 19,717 responses!

# Prepare the data analysis
Before starting the analysis, we need to make few preparation: load the packages, load and inspect the data.

# Load packages
We load the packages used for the analysis.

```python
import sys
import os
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
color = sns.color_palette()
%matplotlib inline

```

# Load the data
Let's see first what data files do we have in the root directo

There are four files with the results of the survey.
1 - multipleChoiceResponses.csv - Multiple choice single response questions fit into individual columns whereas multiple choice multiple response questions were split into multiple columns. Text responses were encoded to protect user privacy and countries with fewer than 50 respondents were grouped into the category "other".
9
2- Surveyschema.csv - Survey schema describing which questions were presented to which respondents. In general, respondents with more experience were asked more questions than respondents with less experience.

3 - other_text_responses.csv - If "Other" is selected there is also an option to provide a text response. These text responses were separated and shuffled to protect user privacy.

4 - questions_only.csv - The list of questions from the 2019 Kaggle Data Science and Machine Learning Survey

Let us start our analysis by looking at the data from from multipleChoiceResponses.csv file

```python
#Importing the 2019 Dataset
df_2019 = pd.read_csv('../input/kaggle-survey-2019/multiple_choice_responses.csv')
df_2019.columns = df_2019.iloc[0]
df_2019=df_2019.drop([0])
```

```python
df_2019.head(2)
```

```python
import matplotlib.pyplot as plt
nationality=df_2019['In which country do you currently reside?'].value_counts()
print(nationality)
```

# Country 
The largest number of responses to survey were from India and Us. People from India participate in survey mostly.

```python

plt.figure(figsize=(20,60))
ax=df_2019['In which country do you currently reside?'].value_counts()[:59].plot.barh(width=0.6,color=sns.color_palette('Set2',25))
plt.gca().invert_yaxis()
plt.title('Country')
plt.show()
```

# AGE -
As picture speaks itself Millennials dominate data science, with 25-29 year olds being the most common age bracket.


```python
df_2019.iloc[1:, :]['What is your age (# years)?'].value_counts().plot(kind='bar', figsize=(20, 8))
plt.xlabel('Age Groups')
plt.ylabel('Count')
plt.title('Age Groups Distribution')
plt.show()

```
