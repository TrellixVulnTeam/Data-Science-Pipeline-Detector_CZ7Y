```python
def merge_col_name(df):
    new_cols = []
    for i in range(len(df.columns.to_list())):
        new_cols.append(' '.join(df.columns.to_list()[i]))

    df.columns = new_cols
```

```python
# rename countries to match shapefile and perform further cleanup 
def modify_country(df):
    df['country'] = df['country'].apply(
        lambda x: 'Iran' if x=='Iran, Islamic Republic of...'
        else 'United States' if x=='United States of America'
        else 'United Kingdom' if x=='United Kingdom of Great Britain and Northern Ireland'
        else 'Hong Kong' if x=='Hong Kong (S.A.R.)'
        else 'Undisclosed' if x=='I do not wish to disclose my location'
        else x
    )

modify_country(df_19)
modify_country(df_20)
modify_country(df_21)
```

```python
def break_range(x, df):
    df[x].fillna('-', inplace=True)
    df[x] = df[x].str.replace(r'[^\d-]', '', regex=True)

    new_col = x + '_avg'
    for i in range(df.shape[0]):
        if(df[x][i] == '-'):
            df.at[i,new_col] = np.NaN
        elif(df[x][i].find('-') != -1):
            min_year = list(map(int, df[x][i].split('-')))[0]
            max_year = list(map(int, df[x][i].split('-')))[1]
            df.at[i,new_col] = (min_year + max_year)/2
        elif(df[x][i]==''):
            df.at[i,new_col] = 0
        else:
            df.at[i,new_col] = int(df[x][i])
```

```python
for df in [df_19, df_20, df_21]:    
    break_range('age_range', df)
    break_range('yearly_salary', df)
    break_range('coding_years', df)
    break_range('ml_years', df)
    break_range('company_size', df)
    break_range('data_science_workers', df)
    break_range('ml_spending', df)
```

```python
# check percentage of nulls in all columns and drop those with 50% null values
def drop_50_null(df):
    nulls = (df.isna().sum()/df.shape[0]).reset_index(name='perc')
    nulls_todrop = nulls[nulls['perc']>=0.5]['index']

    df.drop(nulls_todrop, axis=1, inplace=True)
    
drop_50_null(df_19)
drop_50_null(df_20)
drop_50_null(df_21)
```
