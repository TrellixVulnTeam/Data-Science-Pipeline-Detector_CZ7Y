```python
import numpy as np
import pandas as pd 

import utils_data_prepping as udp
import utils_clf_models as clf
import sklearn.metrics as skm

import seaborn as sns
sns.set(rc={'figure.figsize':(12,10)})
import matplotlib.pyplot as plt
%matplotlib inline

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
train_df = udp.loading('/kaggle/input/otto-group-product-classification-challenge/train.csv')
test_df = udp.loading('/kaggle/input/otto-group-product-classification-challenge/test.csv')
```

```python
sns.histplot(train_df['target'])
plt.show()
```

# Modeling Plans
* 3 single Models: Log Reg, Tree, K-NN
* 2 Ensemble Models: Random Forest & XGBoost
* Compare model results 
* Hyperparamter Tuning

```python
# Training dataframe
X, y = udp.pre_processing(train_df, 'target')
clf1 = clf.Classifier(X, y, 'knn')
clf1.preprocess_split(1/3, 1)
clf1.fit_predict()
print('For the training set:')
clf1.metrics(printing=True)

# Testing Dataset
new_test = test_df.to_numpy()[:y.shape[0]]
clf2 = clf.Classifier(new_test, y, 'knn')
clf2.preprocess_split(1/3, 8)
clf2.fit_predict()
print('For the testing set:')
clf2.metrics(printing=True)
```

```python
# Training dataframe
X, y = udp.pre_processing(train_df, 'target')
clf1 = clf.Classifier(X, y, 'log_reg')
clf1.preprocess_split(1/3, 1)
clf1.fit_predict({'multi_class': 'ovr'})
print('For the training set:')
clf1.metrics(printing=True)

# Testing Dataset
new_test = test_df.to_numpy()[:y.shape[0]]
clf2 = clf.Classifier(new_test, y, 'log_reg')
clf2.preprocess_split(1/3, 8)
clf2.fit_predict({'multi_class': 'ovr'})
print('For the testing set:')
clf2.metrics(printing=True)
```

```python
# Training dataframe
X, y = udp.pre_processing(train_df, 'target')
clf1 = clf.Classifier(X, y, 'tree')
clf1.preprocess_split(1/3, 1)
clf1.fit_predict({'multi_class': 'ovr'})
print('For the training set:')
clf1.metrics(printing=True)

# Testing Dataset
new_test = test_df.to_numpy()[:y.shape[0]]
clf2 = clf.Classifier(new_test, y, 'tree')
clf2.preprocess_split(1/3, 8)
clf2.fit_predict({'multi_class': 'ovr'})
print('For the testing set:')
clf2.metrics(printing=True)
```
