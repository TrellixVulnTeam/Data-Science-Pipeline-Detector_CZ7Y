```python
# Reading the train, test data
train = pd.read_csv("/kaggle/input/otto-group-product-classification-challenge/train.csv")
test = pd.read_csv("/kaggle/input/otto-group-product-classification-challenge/test.csv")
sample = pd.read_csv("/kaggle/input/otto-group-product-classification-challenge/sampleSubmission.csv")

print(train.shape, test.shape, sample.shape)
```

```python
import seaborn as sns
sns.countplot(train.target)
```

```python
X = train.drop(['id','target'],axis = 1)
y = train['target']

X_test1 = test.drop(['id'],axis = 1)

print(X.shape, y.shape, X_test1.shape)
```

```python
# import libraries
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV, RandomizedSearchCV
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
```

```python
# Encoding the target variable
result = LabelEncoder()
y = result.fit_transform(y)
```

```python
from collections import Counter
Counter(y)

# Class imbalance 
```

```python
# keeping 20% dataset for validation
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .2)
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier().fit(X_train, y_train)
y_pred = model.predict_proba(X_test)
from sklearn.metrics import log_loss

print(log_loss(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, model.predict(X_test)))
```

```python
import xgboost as xgb
model2 = xgb.XGBClassifier().fit(X_train,y_train)
y_pred = model2.predict_proba(X_test)
print(log_loss(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, model2.predict(X_test)))
```

```python
# Fitting on entire train dataset
model3 = xgb.XGBClassifier().fit(X,y)
```

```python
pred = model3.predict_proba(X_test1)
submission_df = pd.DataFrame(pred,columns = ['Class_1','Class_2','Class_3','Class_4','Class_5','Class_6',
                                            'Class_7','Class_8','Class_9'], index = X_test1.index)
submission_df['id'] = test['id']
```

```python
submission_df.to_csv('subs.csv',index = False)
```
