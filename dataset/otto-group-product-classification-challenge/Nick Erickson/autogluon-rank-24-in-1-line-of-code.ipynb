```python
# Load the data
from autogluon.tabular import TabularDataset, TabularPredictor
train_data = TabularDataset('../input/otto-group-product-classification-challenge/train.csv')
test_data = TabularDataset('../input/otto-group-product-classification-challenge/test.csv')
```

```python
# Specify the label column, eval metric, time limit
label = 'target'
eval_metric = 'log_loss'
time_limit = 3600 * 6  # 6 hours
```

```python
# Fit end-to-end with raw data in one line of code
predictor = TabularPredictor(
    label=label, eval_metric=eval_metric, learner_kwargs={'ignored_columns': ['id']}
).fit(train_data, presets='best_quality', time_limit=time_limit)
```

```python
predictor.leaderboard(silent=True)
```

```python
import pandas as pd
submission = test_data[['id']]
test_pred_proba = predictor.predict_proba(test_data)
submission = pd.concat([submission, test_pred_proba], axis=1)
submission.to_csv('submission.csv', index=False)
submission.head()
```
