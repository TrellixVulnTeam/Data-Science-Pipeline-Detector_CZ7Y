```python
import pandas as pd
import numpy as np
from sklearn import ensemble , feature_extraction , preprocessing

train = pd.read_csv("../input/otto-group-product-classification-challenge/train.csv")
test = pd.read_csv("../input/otto-group-product-classification-challenge/test.csv")
sample = pd.read_csv("../input/otto-group-product-classification-challenge/sampleSubmission.csv")

```

```python
train.head()
```

```python
labels = train.target.values
train = train.drop('id',axis = 1)
train = train.drop('target',axis = 1)
test = test.drop('id',axis = 1)
```

```python
lbl_enc = preprocessing.LabelEncoder()
labels = lbl_enc.fit_transform(labels)
```

```python
clf = ensemble.RandomForestClassifier(n_jobs=-1, n_estimators=100, max_features = 50, verbose=2)
clf.fit(train ,labels)
```

```python
sample.columns
```

```python
preds = clf.predict_proba(test)
```

```python
preds = pd.DataFrame(preds , index=sample.id.values , columns=sample.columns[1:])
```

```python
preds.to_csv('benchmark.csv', index_label='id')
```
