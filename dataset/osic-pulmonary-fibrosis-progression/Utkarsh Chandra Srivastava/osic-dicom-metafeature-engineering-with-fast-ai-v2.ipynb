## Generate DICOM Metadata DataFrame with fast.ai v2

With much thanks to fast.ai's medical module we can easily generate metadata from DICOM Files with very few  lines of code 

Based on this great kernel https://www.kaggle.com/jhoward/creating-a-metadata-dataframe-fastai/notebookgreat kernel by Sir Jeremy Howard

```python
from fastai2.basics           import *
from fastai2.medical.imaging  import *
```

```python
main_dir=Path("../input/osic-pulmonary-fibrosis-progression")
```

```python
files=[]
for dirname, _, filenames in os.walk(main_dir):
    for filename in (filenames):
        files.append(os.path.join((dirname), filename))        
files=[x for x in files if '.csv' not in x]
train_images= [Path(x) for x in files if 'train'  in x]
test_images = [Path(x) for x in files if 'test'   in x]
```

To turn the DICOM file metadata into a DataFrame we can use the from_dicoms function that fastai v2 adds. By passing px_summ=True summary statistics of the image pixels (mean/min/max/std) will be added to the DataFrame as well.


```python
dcm_metadata_train=pd.DataFrame.from_dicoms(train_images,px_summ=True)
dcm_metadata_test=pd.DataFrame.from_dicoms(test_images,px_summ=True)
```

```python
dcm_metadata_train=dcm_metadata_train.dropna(axis=1)
dcm_metadata_train=dcm_metadata_train.drop("PatientSex",axis=1)
dcm_metadata_test=dcm_metadata_test.dropna(axis=1)
dcm_metadata_test=dcm_metadata_test.drop("PatientSex",axis=1)
```

```python
dcm_metadata_train.to_csv("dcm_metadata_train.csv")
dcm_metadata_test.to_csv("dcm_metadata_test.csv")
```
