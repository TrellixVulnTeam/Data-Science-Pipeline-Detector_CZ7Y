### Packages

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

from os import listdir

from fastai.basics import *
from fastai.medical.imaging import *

print('Done!')
```

```python
base_dir = '../input/osic-pulmonary-fibrosis-progression'
#fast.ai is required to create a Path object, which is read using the from_dicoms method into the df
base_path = Path('../input/osic-pulmonary-fibrosis-progression')


train_sample_path = base_path/'train'
fns_trn = train_sample_path.ls()
```

```python
def read_dcms(path):
    file_dcms = path.ls()
    #You can change px_summ to true for some additional information that is useful for finding trends 
    #and corrupeted files, but I was already having trouble getting through all of the DICOMS without
    #going over Kaggle's kernel RAM limit.
    dcm_df = pd.DataFrame.from_dicoms(file_dcms, px_summ=False)
    return dcm_df
```

```python
full_dcm_df = pd.DataFrame()

for folder in fns_trn:
    patient_df = read_dcms(folder)
    full_dcm_df = full_dcm_df.append(patient_df, ignore_index=True)
```

There's a lot of images (over 33,000!) which means there's a lot of metadata.

```python
full_dcm_df.shape
```

 We've got quite a few different types of metadata as well. Some are not going to be useful for this competition, but pixel spacing, slice thickness, and a few others will be especially usefull. Also, be aware: not all columns are used for every image.
