```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)


```

```python
import pydicom
import os
import matplotlib.pyplot as plt
```

```python
from glob import glob
```

```python
patient="../input/osic-pulmonary-fibrosis-progression/train/ID00007637202177411956430/"
```

```python
scans=glob(patient+"/*.dcm")
```

```python
def load(path):
    scans=os.listdir(path)
    slices=[]
    for s in scans:
        with pydicom.dcmread(path+"/"+s) as f:
            slices.append(f)
    slices.sort(key = lambda x: int(x.InstanceNumber))
    
    return slices
```

```python
p=load(patient)
```

```python
def getImage(scans):
    image = np.stack([s.pixel_array for s in scans])
    image = image.astype(np.int16)
    
    image[image<=-1000] = 0
    
    return np.array(image, dtype=np.int16)
imgs= getImage(p)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 10))
ax[0].set_title("original")
ax[0].imshow(p[25].pixel_array)
ax[1].set_title("HU image")
ax[1].imshow(imgs[25])
```
