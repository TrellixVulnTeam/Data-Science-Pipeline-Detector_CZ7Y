## Exploratory Data Analysis (EDA)
Exploratory Data Analysis (EDA) is the process of exploring a dataset and gaining an understanding of its main characteristics. The dataprep.eda package makes this process easier by allowing users to explore important characteristics using simple APIs. Each API enables the user to analyse the dataset at various levels, from high to low, and from various perspectives.
