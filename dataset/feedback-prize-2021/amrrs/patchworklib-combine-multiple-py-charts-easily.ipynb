```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import patchworklib as pw
```

```python
train = pd.read_csv("/kaggle/input/feedback-prize-2021/train.csv")
```

```python
train.head()
```

```python
train['discourse_duration'] = train['discourse_end'] - train['discourse_start']
```

```python
train.head()
```

### 1st step = Create a Brick Object

```python
ax1 = pw.Brick("ax1", figsize = (3,2))
```

```python
sns.scatterplot(data = train, x = 'discourse_end', y = 'discourse_start', ax = ax1)
```

```python
ax1.set_title("Relationship b/w Discourse Start and End")
```

```python
ax2 = pw.Brick("ax2", figsize = (4,2))
```

```python
sns.histplot(data = train, x = 'discourse_duration', ax = ax2)
```

```python
ax2.set_title("Distribution of Discourse Duration")
```

```python
ax3 = pw.Brick("ax3", figsize = (4,3))
```

```python
sns.barplot(data = train, y = 'discourse_type', x = 'discourse_duration', ax = ax3)
```

```python
ax3.set_title("Discourse type and their duration")
```

Stack Charts Horizontally

```python
ax_h123 = ax1 | ax2 | ax3
```

```python
ax_h123.savefig()
```

Stack the charts vertically

```python
ax_v123 = ax1 / ax2 / ax3
```

```python
ax_v123.savefig()
```

```python
ax_v3_12h = ax3 | ax1 / ax2
```

```python
ax_v3_12h.savefig()
```

```python
ax_v3_12v = ax3 / ax2 | ax1
```

```python
ax_v3_12v.savefig()
```
