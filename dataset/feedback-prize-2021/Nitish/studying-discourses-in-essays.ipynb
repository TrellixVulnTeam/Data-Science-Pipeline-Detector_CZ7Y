### ***Further, we'll look at the correlation among all dicourses***

```python
discourse_count_df = pd.DataFrame(columns=['id', 'Lead', 'Position', 'Evidence', 'Claim', 'Concluding Statement', 'Counterclaim', 'Rebuttal'])

# prepping train_text_df
for group_id, group_df in train_df.groupby('id'):
    discourse_dict = group_df.discourse_type.value_counts().to_dict()
    discourse_dict['id'] = group_id
    discourse_count_df = discourse_count_df.append(discourse_dict, ignore_index=True)

discourse_count_df = discourse_count_df.fillna(0)
train_text_df = train_text_df.merge(discourse_count_df, how='left', on='id')

plt.figure(figsize=(9, 7))
plt.title('Correlation among all discourses')
corr_df = train_text_df.drop(['id', 'text', 'sentence_count', 'word_count'], axis=1).corr(method='pearson')
mask = np.triu(np.ones_like(corr_df))
hmap=sns.heatmap(corr_df,cmap="Spectral")
plt.show()
```

### ***Takeaways***
### ***- The highest correlation is between Rebuttal(an opposing argument or debate) and Counterclaim which maskes sense.***
### ***- The next highest correlation was between Claim and Evidence - the 2 most common discourses.***
### ***- Looking at the correlation between character count and all features, Evidence, Claim and Lead tend to occur the most when it is a bigger essay.***

### ***Next, let's try to answer the question - If all the discourses were to be present in the sentence, how much would each discourse make up in the sentence?***

```python
train_df['len_discourse'] = train_df['discourse_end'] - train_df['discourse_start']
disc_prop_dict = dict((k, []) for k in ['Lead', 'Position', 'Evidence', 'Claim', 'Concluding Statement', 'Counterclaim', 'Rebuttal'])

for group_id, group_df in train_df.groupby('id'):
    # for each essay
    # get length of each essay
    essay_len = int(train_text_df[train_text_df['id']==group_id]['char_count'].iloc[0])
    char_count = train_text_df[train_text_df['id']==group_id].char_count.iloc[0]
    
    group_df['discourse_proportion'] = group_df['len_discourse']/char_count
    disc_agg_series = group_df.groupby('discourse_type')['discourse_proportion'].agg('sum').round(2)
    for discourse, discourse_proportion in disc_agg_series.items():
        disc_prop_dict[discourse].append(discourse_proportion)
```

```python
new_disc_prop_dict = {}
for disc, arr in disc_prop_dict.items():
    new_disc_prop_dict[disc] = round(sum(arr)/len(arr), 2)
new_disc_prop_dict = dict(sorted(new_disc_prop_dict.items(), key=lambda kv: kv[1]))

plt.figure(figsize=(16,9))
plt.title('Discourse propotion in essays')
colors = sns.color_palette("Paired")
plt.pie(x= new_disc_prop_dict.values(), labels = new_disc_prop_dict.keys(), colors = colors,
        explode= [0.1]*len(new_disc_prop_dict.keys()) ,autopct='%.0f%%', shadow = True)
plt.show()
```

### ***The graph shows the proportion of discourses if they are present in the essay. Evidence takes the largest proportion making upto almost half the essay. Claim, concluding statement and lead make up more or less similar percentages in an essay.***

### ***Alright, now let's see where in an essay are the discourses most likey to be found.***

```python
start_proportion = []
for i,row in train_df.iterrows():
    essay_len = int(train_text_df[train_text_df['id']==row.id]['char_count'].iloc[0])
    start_proportion.append(row.discourse_start/essay_len)

train_df['start_proportion'] = start_proportion

disc_start_proportion = {}
for discourse, disc_group_df in train_df.groupby('discourse_type'):
    disc_start_proportion[discourse] = disc_group_df.start_proportion.to_list()
```

```python
reorder_list = ['Lead', 'Position', 'Claim', 'Evidence',  'Counterclaim', 'Rebuttal',  'Concluding Statement']
disc_start_proportion = {k: disc_start_proportion[k] for k in reorder_list}
labels, data = disc_start_proportion.keys(), disc_start_proportion.values()

plt.figure(figsize=(14,9))
plt.title('Distribution of discourses in essays')
plt.boxplot(data, notch=True, patch_artist=True,  boxprops=dict(facecolor="#0096FF"))
plt.xticks(range(1, len(labels) + 1), labels)
plt.show()
```

### ***This plot depicts the distribution of start position of a discourse in an essay with start to end of an essay being 0 to 1.***

### - ***Lead ususally is the first thing you would see in an essay and probably explains what the essay is going to be about.*** 
### - ***Position is the overall stance that the writer of an essay takes in answering the question or on the issue/topic at hand. The meadian from it's boxplot depicts it usually follows Lead but with so many outliers it also means it can be distributed throughout the essay.***
### - ***Next, the essay structure is followed by Claim, Evidence, Counterclaim and Rebuttal.***
### - ***Concuding statement is often the last sentence in an essay but there are quite a few conclusions made along the way as well.***

### ***Cool, hoomans make mistakes***. ***Let's find out if there are spelling mistakes in the essays***

```python
spell = SpellChecker()

def spell_check(text):
    word_list=text.split()
    incorrect_word_count = len(list(spell.unknown(word_list)))
    return incorrect_word_count

incorrect_spelling_df = pd.DataFrame(columns=['total_word_count', 'incorrect_word_count'])
for i, row in train_text_df.sample(n = 1000, replace = False).iterrows():
    incorrect_spelling_df = incorrect_spelling_df.append({
        "total_word_count": row.word_count,
        "incorrect_word_count": spell_check(row.text)
    }, ignore_index=True)

plt.figure(figsize=(12,7))
plt.title('Total words vs incorrect word count in essays')
sns.scatterplot(data = incorrect_spelling_df, x = "incorrect_word_count", y = "total_word_count")
plt.show()
```

### ***There are good few spelling mistakes in the essays. The above graph is from sample of 1000 text files.***
### ***A simple scatter plot shows the obvious relation between number number of incorrect words and total words in the essays. There are words in the essays such as Generic_name or student_name that are also counted in this. Well, the point is we'll have to correct these mistakes before processing any text.***

### ***Finally, let's look at some sample texts in different colours to understand the structure.***

```python
# code credits: https://www.kaggle.com/ilialar/feedback-prize-simple-eda

def color_text(text_id, train_df, color_scheme = None):
    if not color_scheme:
        color_scheme = {
        'Lead': 'red',
        'Position': 'green',
        'Claim': 'blue',
        'Counterclaim': 'magenta',
        'Rebuttal': 'yellow',
        'Evidence': 'cyan',
        'Concluding Statement': 'yellow'
    } 
    with open(f'/kaggle/input/feedback-prize-2021/train/{text_id}.txt') as f:
        lines = f.readlines()
    text = ''.join(lines)
    
    annot_df = train_df[train_df.id == text_id]
    blocks = [(int(row['discourse_start']),int(row['discourse_end']), color_scheme[row['discourse_type']]) for k, row in annot_df.iterrows()]
    blocks.sort()
    i = 0
    last_symbol = -1
    while i < len(blocks):
        if blocks[i][0] > last_symbol + 1:
            blocks.insert(i, (last_symbol+1, blocks[i][0] - 1, None))
        last_symbol = blocks[i][1]
        i += 1
    if last_symbol < len(text):
        blocks.append((last_symbol+1, len(text) - 1, None))

    colored_text = ''.join([colored(text[x[0]:x[1]+1], x[2]) for x in blocks])
    return colored_text
```
