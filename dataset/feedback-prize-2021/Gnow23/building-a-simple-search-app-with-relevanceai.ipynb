- quickly search through data
- type in examples 
- preview data in an easily accessible way
- discover new anomalies in the data as they search

```python
from relevanceai import Client 
client = Client()
```

```python
import pandas as pd
ds = client.Dataset("feedback-prize-2021")
df = pd.read_csv("../input/feedback-prize-2021/train.csv")
ds.insert_df(df)
```

```python
ds.launch_search_app()
```

And that's it! You can configure the search application easily by editing the configuration in the top right!
