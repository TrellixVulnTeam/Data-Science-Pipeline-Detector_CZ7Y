```python
import pandas as pd
pd.set_option("display.precision", 12)
```

```python
train_data_path = "../input/feedback-prize-2021/train.csv"
train_data = pd.read_csv(train_data_path)
train_data.head()
```

```python
discourse_info = train_data.loc[:, ['id', 'discourse_start',
                                    'discourse_type', 'discourse_type_num']]
```

```python
check_data = discourse_info.groupby('id').__iter__()
```

```python
_, g_data = next(check_data)
g_data
```

```python
_, g_data = next(check_data)
g_data
```

```python
%%time
id_list = []
types_list = []
uniq_types_list = []

for name, group in discourse_info.groupby('id'):
    sorted_group = group.sort_values(by='discourse_start')
    discourse_type = sorted_group['discourse_type']
    
    id_list.append(name)    
    types_list.append(", ".join(discourse_type))
    
    uniq_discourse_type = []
    for i_type in discourse_type:
        if not uniq_discourse_type:
            uniq_discourse_type.append(i_type)

        if uniq_discourse_type and uniq_discourse_type[-1] != i_type:
            uniq_discourse_type.append(i_type)
    
    uniq_types_list.append(", ".join(uniq_discourse_type))
    
discourses_sequence = pd.DataFrame(
    {'types': types_list,
     'uniq_types': uniq_types_list},
            index=id_list
)
```

```python
discourses_sequence
```

```python
%%time
count_data = discourses_sequence['uniq_types'].value_counts()
freq_data = discourses_sequence['uniq_types'].value_counts(normalize=True) \
                                                        .mul(100).round(2) \
                                                        .map("{} %".format)

discourses_sequence_stats = pd.DataFrame(
    {'sequence': count_data.index,
     'count': count_data.values,
     'freq': freq_data.values}    
)
```
