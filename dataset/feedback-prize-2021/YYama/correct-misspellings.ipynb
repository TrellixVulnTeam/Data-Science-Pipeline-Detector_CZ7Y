# autocorrect

```python
! pip3 install autocorrect -qq
from autocorrect import Speller
```

```python
# examples from https://www.kaggle.com/c/feedback-prize-2021/discussion/300620
misspellings = ["conclution",
                "conculsion",
                "coclusion",
                "conclucion"]

spell = Speller(lang='en')
for misspelling in misspellings:
    print('Before: ',misspelling,'\t' ,'After:', spell(misspelling))
```

```python
# load a file from testdata as list
tokenize = False
f = open('../input/feedback-prize-2021/test/0FB0700DAF44.txt', 'r', encoding='UTF-8')
data = f.read()
if tokenize==False:
    words = data.split(' ')
else:
    tokenizer = LongformerTokenizerFast.from_pretrained('allenai/longformer-base-4096')
    encoding = tokenizer.tokenize(data)
    words = []
    for word in encoding:
        words.append(word.replace('Ä ', ''))
```

```python
# correcting misspellings using autocorrect
for word in words:
    cor_word = spell(word)
    if word != cor_word:
        print('Before: ',word,'\t' ,'After:', cor_word)
```

```python
# autocorrect can correct sentences directly
print(spell(data))
```

**Most of words are appropriately corrected, but some words are incorrect.**

# pyspellchecker

```python
! pip3 install pyspellchecker -qq
from spellchecker import SpellChecker
```

```python
spell = SpellChecker()

misspelled = spell.unknown(misspellings)

for misspelling in misspellings:
    print('Before: ',misspelling,'\t','After:', spell.correction(misspelling))
```

```python
for word in words:
    cor_word = spell.correction(word)
    if word != cor_word:
        print('Before: ',word,'\t' ,'After:', cor_word)
```

**If the words have periods or commas, they will be recognized as misspellings, but if you tokenize it and then collect it, there may be no problem.**

```python
! pip install -U textblob -qq
! python -m textblob.download_corpora
from textblob import TextBlob
```

```python
data_ = TextBlob(data)
cor_data = str(data_.correct())

if data != cor_data:
    print("Before: ")
    print(data)
    print('')
    print("After: ")
    print(cor_data)
```

```python
cor_words = cor_data.split(' ')

for word, cor_word in zip(words, cor_words):
    if word != cor_word:
        print('Before: ',word,'\t','After:', cor_word)
```

**Some of the words are corrected appropriately, but many of them are converted to other wrong words.**
