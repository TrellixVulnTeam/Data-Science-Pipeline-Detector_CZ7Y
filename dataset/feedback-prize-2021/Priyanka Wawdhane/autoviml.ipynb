```python
df['discourse_text'] = df['discourse_text'].astype("string")
```

```python
df['discourse_type'] = df['discourse_type'].astype("string")
```

```python
df.info()
```

```python
possible_labels = df.discourse_type.unique()

label_dict = {}

for index, possible_label in enumerate(possible_labels):

    label_dict[possible_label] = index

label_dict
```

```python
df['label'] = df.discourse_type.replace(label_dict)
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(df, test_size=0.25)
```

```python
from autoviml.Auto_NLP import Auto_NLP
```

```python
nlp_column = 'discourse_text'
target = 'label'
```

```python
nlp_column.dtype()
```

```python
nlp_transformer= Auto_NLP(
                nlp_column, train, test, target, score_type='balanced_accuracy',
 modeltype='Classification',top_num_features=50, verbose=2,
 build_model=True)
```
