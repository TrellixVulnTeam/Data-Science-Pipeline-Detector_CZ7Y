```python
# Import usefull package
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import ipywidgets as widgets
from IPython.display import Markdown
import os

path ="//kaggle//input//feedback-prize-2021//"
df = pd.read_csv(path + "train.csv")
```

```python
#df["id"]
#ls = [id for id in df["id"]]
#ls
print("Nos of rows: ", df.shape[0])
print("Nos of cols: ", df.shape[1])
print("Unique ids :", len(df["id"].unique()))
itms= [id for id in df["id"].unique()]
#itms
print("Discourse Type:",df["discourse_type"].unique())
discourse_type = {"Lead":"<span title='Lead' style='color: #000; background:#b3ebff;'>",
                 "Position":"<span title='Position' style='color: #000; background:#fff694;'>",
                 "Evidence":"<span title='Evidence' style='color: #000; background:#9eff91;'>",
                 "Claim": "<span title='Claim' style='color: #000; background:#ffab91;'>",
                 "Concluding Statement":"<span title='Concluding Statement' style='color: #000; background:#9664ed;'>",
                 "Counterclaim":"<span title='Counterclaim' style='color: #000; background:#f794ec;'>",
                 "Rebuttal":"<span title='Rebuttal' style='color: #000; background:#ffb442;'>",
                 "End":"</span>"}

def visualize_text(txt,df:pd.DataFrame):
    formt_txt = ""
    if (df.iloc[0]["discourse_start"] == 0):
        pass
    else:
        formt_txt = txt[0:int(df.iloc[0]["discourse_start"])-1]
    
    for i in range(len(df)):
        formt_txt += "[" + discourse_type[df.iloc[i]["discourse_type"]] + txt[int(df.iloc[i]["discourse_start"]):int(df.iloc[i]["discourse_end"])].replace("\n", "" ) + discourse_type["End"] + "]"
        
    #df.iloc[0]["discourse_start"]-1
    #df.iloc[0]["discourse_end"]
    #df.iloc[0]["discourse_type"]
    
    display (Markdown(formt_txt))
    
    print("---------------Legend:-----------")
    for key in discourse_type.keys():
        display (Markdown(discourse_type[key] + key + discourse_type["End"]))
```

```python
drp = widgets.Dropdown(
    options=itms,#[id for id in df["id"].unique()],
    value=itms[0],
    description='ID :',
)
drp
```

```python
file_path ="//kaggle//input//feedback-prize-2021//train//" + drp.value + ".txt"
f = open(file_path, "r")
#txt = f.read()
#print(txt)

df_t = df[ df["id"] == drp.value]

visualize_text(f.read(),df_t)

df_t
```
