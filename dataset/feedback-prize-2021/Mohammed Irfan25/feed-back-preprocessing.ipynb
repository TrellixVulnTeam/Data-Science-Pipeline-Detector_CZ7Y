```python
import numpy as np
import pandas as pd
from tqdm import tqdm
import os
```

```python
# Load the dataset
df = pd.read_csv('../input/feedback-prize-2021/train.csv')
df.head(5)
```

```python
import ntpath
def path_leaf(path):
    head, tail = ntpath.split(path)
    return tail or ntpath.basename(head)
```

```python
complete_text = {}
path = "../input/feedback-prize-2021/train"
for file_name in os.listdir(path):
    with open(os.path.join(path,file_name)) as file:
        complete_text[path_leaf(file_name).split(".")[0]] = file.read()
complete_text
```

```python
for ind in tqdm(df.index):
    _id = df["id"][ind]
    complete_text[_id] = complete_text[_id].replace(df["discourse_text"][ind],"")
```

```python
complete_text["423A1CA112E2"]
```

```python
# Verification
for ind in tqdm(df.index):
    _id = df["id"][ind]
    discourse_text = df["discourse_text"][ind]
    if(discourse_text in complete_text[_id]):
        print(_id)
```

```python
for key,value in tqdm(complete_text.items()):
    if(value.isspace() or value == ""):
        continue
    else:
        df = df.append({
                        "id":key,
                        "discourse_text":value,
                        "discourse_type":"Other"
                    }, ignore_index = True)
```
