```python
import pandas as pd
```

```python
train = pd.read_csv('../input/feedback-prize-2021/train.csv')
train.head(1)
```

```python
from pathlib import Path

path = Path('../input/feedback-prize-2021/train')

def get_tokens(ids):
    with open(path/f'{ids}.txt', 'r') as file: data = file.read()
    return data.split()

def get_text(ids):
    with open(path/f'{ids}.txt', 'r') as file: data = file.read()
    return data
```

```python
def verify_text(row):
    tokens = get_tokens(row['id'])
    indices = [int(x) for x in row.predictionstring.split()]
    start = indices[0]
    end = indices[-1] + 1
    text = ' '.join(tokens[start:end])
    return text

verify_text(train.loc[120722])
```

```python
train['verify'] = train.apply(verify_text, axis=1)
```

```python
train['correct'] = train.apply(lambda row: ' '.join(row.discourse_text.split()) == row.verify, axis=1)
train.correct.value_counts()
```

```python
train['chardif'] = train.apply(lambda row: len(row.discourse_text) - len(row.verify), axis=1)
```

```python
for _, row in train[train.correct == False].sample(n=5).iterrows():
    print(row['id'])
    print(row.discourse_text)
    print(row.verify)
```

```python
for _, row in train[(train.chardif > 5) & (train.correct == False)].sample(n=5).iterrows():
    print(row['id'])
    print(row.discourse_text)
    print(row.verify)
```
