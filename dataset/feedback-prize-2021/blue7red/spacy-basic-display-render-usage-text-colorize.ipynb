```python
import numpy as np
import pandas as pd

from glob import glob

import spacy
```

# get all train filelist 

```python
train_files = glob("../input/feedback-prize-2021/train/*.txt")
train_files[:5]
```

# get all train file info

```python
train = pd.read_csv("../input/feedback-prize-2021/train.csv")
train.head()
```

# get one sample file contents

```python
sample_file_path = train_files[0]
f = open(sample_file_path, "r")
print(f.read())
```

# get sample file id from file path

```python
sample_file_id = sample_file_path[35:-4] #AFEC37C2D43F
sample_file_id
```

# get sample file info from sample file id 

```python
train.query('id == "' +sample_file_id + '"')
```

```python
train.query('id == "' +sample_file_id + '"')["discourse_type"].value_counts()
```

# colorize contents by discourse_type

```python
ents = []
for i, row in train[train['id'] == sample_file_id].iterrows():
    ents.append({
                    'start': int(row['discourse_start']), 
                     'end': int(row['discourse_end']), 
                     'label': row['discourse_type']
                })

with open(sample_file_path, 'r') as file: data = file.read()

docData = {
    "text": data,
    "ents": ents,
}

colors = {'Lead': '#EE11D0',
          'Position': '#AB4DE1',
          'Claim': '#1EDE71',
          'Evidence': '#33FAFA',
          'Counterclaim': '#4253C1',
          'Concluding Statement': 'yellow',
          'Rebuttal': 'red'}
options = {"ents": train.discourse_type.unique().tolist(), "colors": colors}
spacy.displacy.render(docData, style="ent", options=options, manual=True, jupyter=True);
```
