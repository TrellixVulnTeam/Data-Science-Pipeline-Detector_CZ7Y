```python
# Libraries we need
!pip install bertopic[visualization] --quiet
import pandas as pd
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
import os
import plotly.express as px
```

```python
# Loading Data
x = []
y = []
for i in tqdm(list(os.listdir('../input/feedback-prize-2021/train'))):
    x.append(i.replace('.txt',''))
    y.append(open('../input/feedback-prize-2021/train/' + i, 'r').read())
    df = pd.DataFrame({'id': x, 'text': y})
df.head()
```

```python
# visualising text Length
temp1 = [len(i.split()) for i in list(df['text'])]
x1 = list(set(temp1))
y1 = [temp1.count(i) for i in x1]
fig = px.bar(x=x1, y=y1,title='Distribution of Sentence Length',color_discrete_sequence=['black']*3)
fig.show()
```

```python
# creating some extra feature so we can visualize in more detail
df['count_chars'] = df['text'].apply(lambda x:len(x))

# count capital char
def count_capital_chars(text):
    count=0
    for i in text:
        if i.isupper():
            count+=1
    return count
df['capital_char'] = df['text'].apply(lambda x:count_capital_chars(x))

# count capital words
def count_capital_words(text):
    return sum(map(str.isupper,text.split()))

df['captial_words'] = df['text'].apply(lambda x:count_capital_words(x))

# Count the number of punctuations
def count_punctuations(text):
    punctuations="!#$%&'()*+,-./:;<=>?@[\]^_`{|}~"
    d = 0
    for i in punctuations:
        d+=1
    return d 

df['no_of_punctuations'] = df['text'].apply(lambda x:count_punctuations(x))

# count number of  words
df['word_count'] = df['text'].apply(lambda x:len(x.split()))

# unique words
df['unique_word'] = df['text'].apply(lambda x:len(list(set(x.split()))))

# number of sentence
import nltk
def count_sent(text):
    return len(nltk.sent_tokenize(text))

df['no_of_sent'] = df['text'].apply(lambda x:count_sent(x))

# Calculating average word length
df['avg_word_len'] = df['count_chars']/df['word_count']

#  unique words vs word count feature
df['unique_vs_word'] = df['unique_word']/df['word_count']

# count number of stop_words
def count_stopwords(text):
    from nltk.corpus import stopwords
    from nltk.tokenize import word_tokenize
    stop_words = set(stopwords.words('english'))  
    word_tokens = word_tokenize(text)
    stopwords_x = [w for w in word_tokens if w in stop_words]
    return len(stopwords_x)
df['stop_words'] = df['text'].apply(lambda x:count_stopwords(x))
```
