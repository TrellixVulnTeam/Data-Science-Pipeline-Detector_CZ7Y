## Evaluation

Submissions are evaluated on the overlap between ground truth and predicted word indices.  
投稿された単語は、グランドトゥルースと予測された単語のインデックスの重なり具合で評価されます。

1. For each sample, all ground truths and predictions for a given class are compared.
2. If the overlap between the ground truth and prediction is >= 0.5, and the overlap between the prediction and the ground truth >= 0.5, the prediction is a match and considered a true positive. If multiple matches exist, the match with the highest pair of overlaps is taken.
3. Any unmatched ground truths are false negatives and any unmatched predictions are false positives.  


1. 各サンプルについて、与えられたクラスに関するすべてのグランドトゥルースと予測値が比較される。
2. グランドトゥルースと予測の重なりが >= 0.5 で、予測とグランドトゥルースの重なりが >= 0.5 の場合、予測はマッチし、真陽性とみなされます。複数のマッチが存在する場合、最も高いオーバーラップの組を持つマッチが選ばれる。
3. マッチしていないグランドトゥルースは偽陰性、マッチしていない予測は偽陽性となります。

例

Ground Truth
```
id,class,predictionstring
1,Claim,1 2 3 4 5
1,Claim,6 7 8
1,Claim,21 22 23 24 25
```

Predition
```
id,class,predictionstring
1,Claim,1 2
1,Claim,6 7 8
```

The first prediction would not have >= 0.5 overlap with either ground truth and would be a false positive. The second prediction would overlap perfectly with the second ground truth and be a true positive. The third ground truth would be unmatched, and would be a false negative.  
最初の予測は、どちらのグランドトゥルースとも0.5以上のオーバーラップがなく、偽陽性となるであろう。2番目の予測は2番目のグランドトゥルースと完全にオーバーラップし、トゥルーポジティブとなる。3番目のGround Truthはマッチングせず、偽陰性となる。

The final score is arrived at by calculating TP/FP/FN for each class, then taking the macro F1 score across all classes.  
最終的なスコアは、各クラスのTP/FP/FNを計算し、全クラスのmacro F1 scoreを取ることで算出される。

The word indices are calculated by using Python's .split() function and taking the indices in the resulting list. The two overlaps are calculated by taking the set() of each list of indices in a ground truth / prediction pair and calculating the intersection between the two sets divided by the length of each set.  
単語のインデックスはPythonの.split()関数を使用し、結果のリストのインデックスを取ることで計算しています。2つのオーバーラップは、グランドトゥルース/予測ペアのインデックスの各リストのset()を取り、各セットの長さで割った2つのセットの交点を計算することにより算出されます。

Your task is to predict the human annotations. You will first need to segment each essay into discrete rhetorical and argumentative elements (i.e., discourse elements) and then classify each element as one of the following:  
課題は、人間のアノテーションを予測することです。あなたはまず、各エッセイを個別の修辞的および議論的要素（すなわち、談話要素）に区分し、次に各要素を以下のいずれかに分類する必要があります。

- Lead - an introduction that begins with a statistic, a quotation, a description, or some other device to grab the reader’s attention and point toward the thesis
- Position - an opinion or conclusion on the main question
- Claim - a claim that supports the position
- Counterclaim - a claim that refutes another claim or gives an opposing reason to the position
- Rebuttal - a claim that refutes a counterclaim
- Evidence - ideas or examples that support claims, counterclaims, or rebuttals.
- Concluding Statement - a concluding statement that restates the claims


- リード - 統計、引用、説明などで始まる序文で、読者の注意を引き、論文への方向を示す。
- ポジション - 本題に対する意見または結論
- 主張 - ポジションを支持する主張
- 反証 - 他の主張に反論する、または立場に反対する理由を示す主張。
- 反論 - 反証に反論する主張
- 証拠 - 主張、反論、または反証をサポートするアイデアや例。
- 総括 - 主張を再表現する結論となる文。

````python
import re

def atoi(text):
    return int(text) if text.isdigit() else text

def natural_keys(text):
    return [ atoi(c) for c in re.split(r'(\d+)', text) ]

discourse_type_num_classes = sorted(np.unique(train["discourse_type_num"]), key=natural_keys)
````

````python
fig = px.bar(
    x=discourse_type_num_classes,
    y=[
        list(train["discourse_type_num"]).count(i)
        for i in discourse_type_num_classes
    ],
    color = discourse_type_num_classes,
    color_continuous_scale="blues") 
fig.update_xaxes(title="Classes")
fig.update_yaxes(title = "Number of Rows")
fig.update_layout(showlegend = True,
    title = {
        'text': 'Enumerated class label of Discourse Element Distribution ',
        'y':0.95,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'},
        template="plotly_white")
fig.show()
````

## Text Visualization

````python
for r in [20, 100, 500, 2000, 5000]:
    print(f'============== {r} ==================')
    ents = []
    for i, row in train[train['id'] == train_files[r][35:-4]].iterrows():
        ents.append({
                        'start': int(row['discourse_start']), 
                         'end': int(row['discourse_end']), 
                         'label': row['discourse_type']
                    })

    with open(train_files[r], 'r') as file: data = file.read()

    doc2 = {
        "text": data,
        "ents": ents,
    }

    colors = {'Lead': '#EE11D0','Position': '#AB4DE1','Claim': '#1EDE71','Evidence': '#33FAFA','Counterclaim': '#4253C1','Concluding Statement': 'yellow','Rebuttal': 'red'}
    options = {"ents": train.discourse_type.unique().tolist(), "colors": colors}
    spacy.displacy.render(doc2, style="ent", options=options, manual=True, jupyter=True)
    print('\n\n')
````
