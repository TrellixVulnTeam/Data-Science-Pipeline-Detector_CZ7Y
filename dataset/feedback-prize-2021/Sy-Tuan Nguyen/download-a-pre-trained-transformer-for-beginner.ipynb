This competition requires internet OFF, this notebook shows 'beginners' how to download a pre-trained model and save it to a dataset for use with internet OFF.

```python
import os
from transformers import AutoTokenizer, AutoConfig, TFAutoModel
```

```python
# Download pre-trained model, with internet ON
# List of pre-trained models can be found via this link: https://huggingface.co/transformers/v3.3.1/pretrained_models.html
MODEL = 'allenai/longformer-base-4096'

tokenizer = AutoTokenizer.from_pretrained(MODEL)
config = AutoConfig.from_pretrained(MODEL) 
backbone = TFAutoModel.from_pretrained(MODEL, config=config)
```

```python
# Save downloaded model to a dataset that can be used after without internet connection
DIR = 'transformer_model'
os.mkdir(DIR)
tokenizer.save_pretrained(DIR)
config.save_pretrained(DIR)
backbone.save_pretrained(DIR)
```
