```python
def make_wordcloud(sent):
    stop_words = set(["one","little","said","the","they","that","you","in","of","and","those",
                     "or","some","when","what","where","why","how","be","to","so","as","is",
                      "am","are","was","were","this","it","on","not","no","their","all",
                     "by","upon","their","them","its","with","for","but","at","hi","there",
                     "into","from","him","his","her","such","he","she","which","an",
                     "out","then","can","up","would","if","we","about","did","my","will",
                      "very","other","could","our","thing","over","under","because","also",
                     "should","like","able","say","have","do"])
    wordcloud = WordCloud(background_color="white", width=500, height=500,prefer_horizontal=1,stopwords=stop_words,
                         regexp=r"\w[\w]+").generate(sent)
    return wordcloud
id_list = [df.iloc[0:6583]["id"].unique(),
           df.iloc[6584:14961]["id"].unique(),
           df.iloc[14962:24806]["id"].unique(),
           df.iloc[24807:29964]["id"].unique(),
           df.iloc[29965:38419]["id"].unique(),
           df.iloc[38420:45671]["id"].unique(),
           df.iloc[45672:54846]["id"].unique(),
           df.iloc[54847:62075]["id"].unique(),
           df.iloc[62076:74658]["id"].unique(),
           df.iloc[74659:87298]["id"].unique(),
           df.iloc[87299:104127]["id"].unique(),
           df.iloc[104128:111201]["id"].unique(),
           df.iloc[111202:129728]["id"].unique(),
           df.iloc[129729:144292]["id"].unique()]
word_clouds = []
#各題材別のwordcloud
for ids in id_list:
    sentences=""
    for txt_id in ids:
        f = open("../input/feedback-prize-2021/train/"+txt_id+".txt","r")
        txt_data = f.read()
        sentences+=txt_data
    word_clouds.append(make_wordcloud(sentences))
titles = ["運転中の電話捜査","カーフリーデー" ,"夏休みの自由研究"
         ,"シーゴーイングカウボーイ","課外活動","金星の探索"
         ,"FACSの教室導入","火星の人面岩","成績不良児の社会奉仕活動"
         ,"自動運転車の開発","アメリカの大統領選挙制度","学校への携帯電話持ち込み"
         ,"リモート授業の導入","アドバイスを聞く人数"]
plt.figure(figsize=(25, 50))
plt.subplots_adjust(wspace=0.2, hspace=0)
for i in range(0,len(id_list)):
    plt.subplot(5, 3, i+1)
    plt.title(str(i+1)+":"+titles[i],fontsize=30)
    plt.imshow(word_clouds[i])
    plt.axis("off")
```

```python
#間に合わせのコードなので真似しないように
def get_discource_by_id(txt_id):
    f = open("../input/feedback-prize-2021/train/"+txt_id+".txt","r")
    txt_data = f.read()
    Lead = df[(df["id"]==txt_id)&(df["discourse_type"]=="Lead")]
    Position = df[(df["id"]==txt_id)&(df["discourse_type"]=="Position")]
    Claim = df[(df["id"]==txt_id)&(df["discourse_type"]=="Claim")]
    CounterClaim = df[(df["id"]==txt_id)&(df["discourse_type"]=="Counterclaim")]
    Rebuttal = df[(df["id"]==txt_id)&(df["discourse_type"]=="Rebuttal")]
    Evidence = df[(df["id"]==txt_id)&(df["discourse_type"]=="Evidence")]
    ConcludingStatement = df[(df["id"]==txt_id)&(df["discourse_type"]=="Concluding Statement")]
    return txt_data,Lead,Position,Claim,CounterClaim,Rebuttal,Evidence,ConcludingStatement
```
