```python
from lib.feedback_util import read_train_csv, calc_word_indices, stringify_word_indices, get_train_essay_text
```

Now let's use that to check the predictionstring column in 'train.csv'

```python
train = read_train_csv()
```

```python
%%time

train = train.groupby('id').apply(lambda group: group.assign(sanity_check = group.apply(lambda row: stringify_word_indices(
    calc_word_indices(get_train_essay_text(group.name), row.discourse_start, row.discourse_end)
) == row.predictionstring, axis = 1)))
```
