**Predictionstring from char indices**

Below is the algorithm published by @aignerpicou in [this thread](https://www.kaggle.com/c/feedback-prize-2021/discussion/297688)

```python
def calc_word_indices(full_text, discourse_start, discourse_end):
    start_index = len(full_text[:discourse_start].split())
    token_len = len(full_text[discourse_start:discourse_end].split())
    output = list(range(start_index, start_index + token_len))
    if output[-1] >= len(full_text.split()):
        output = list(range(start_index, start_index + token_len-1))
    return " ".join(str(x) for x in output)
```

Now let's use that to check the predictionstring column in 'train.csv'

```python
import os
import pandas as pd
from tqdm import tqdm
```

```python
train_dir = "../input/feedback-prize-2021/train/"
train_files = os.listdir(train_dir)
train = pd.read_csv("../input/feedback-prize-2021/train.csv",dtype={'discourse_start': int, 'discourse_end': int})
```

```python
for file in tqdm(train_files):
    idx = train['id'] == file[:-4]
    train.loc[idx,'sanity_check'] = train.loc[idx].apply(
        lambda row: calc_word_indices(open(train_dir + file, 'r').read(),
                    row.discourse_start, row.discourse_end) == row.predictionstring,
        axis = 1
    )
```
