```python
import os
import pandas as pd
train_dir = "../input/feedback-prize-2021/train/"
train_files = os.listdir(train_dir)
train = pd.read_csv('../input/feedback-prize-2021/train.csv')
```

**Contradiction in Train data**

From the numbers shown below it seems that:
1. According to the predictionstring field, this discourse ends on the second last word of the text.
2. According to the discourse_end field, the discourse ends on a character in the trailing whitespace of the text.


```python
discourse_id = 1622627628524
row = train.loc[train.discourse_id == discourse_id]
discourse_end = row.discourse_end.values[0]
predictionstring = row.predictionstring.values[0].split()
file_id = row.id.values[0]
text = open(train_dir + file_id + '.txt', 'r').read()

print(int(discourse_end), 'Last character in discourse')
print(len(text.rstrip()), 'Number of useful characters in text, trailing whitespace removed')
print(predictionstring[-1], 'Last word in discourse')
print(len(text.split()), 'Number of words in text')
```
