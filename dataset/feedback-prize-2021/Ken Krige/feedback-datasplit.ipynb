```python
import pandas as pd
import numpy as np
import os
from tqdm.auto import tqdm
from iterstrat.ml_stratifiers import MultilabelStratifiedShuffleSplit
```

```python
df = pd.read_csv("../input/feedback-prize-2021/train.csv")
dfx = pd.get_dummies(df[['id','discourse_type']], columns=["discourse_type"]).groupby(["id"], as_index=False).sum()

# Choose a suitable number of test files here.
splitter = MultilabelStratifiedShuffleSplit(test_size=1000, random_state=56, n_splits=1)
train, val = next(splitter.split(dfx,dfx))

dfx.loc[val,'kfold'] = 0 #'validation'
dfx.loc[train,'kfold'] = 1 # 'train'
df = df.merge(dfx[['id','kfold']], on="id", how="left")

train = df[df.kfold==1].drop('kfold',axis=1)
val = df[df.kfold==0].drop('kfold',axis=1)

df.to_csv("train_folds.csv", index=False)
train.to_csv("train_split.csv", index=False)
val.to_csv("val_split.csv", index=False)
```
