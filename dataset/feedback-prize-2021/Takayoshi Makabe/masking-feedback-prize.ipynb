```python
import random
from transformers import AutoTokenizer


P_MASK = 0.2
```

```python
def get_train_text(ids):
    with open(f'../input/feedback-prize-2021/train/{ids}.txt', 'r') as f:
        text = f.read()
    return text
```

```python
tokenizer = AutoTokenizer.from_pretrained('roberta-base')
```

```python
text = get_train_text('001A03E06F3C')
print(text)
```

```python
token = tokenizer(text)
print(token['input_ids'])
```

```python
token_list = tokenizer.convert_ids_to_tokens(token['input_ids'])
origin_text = tokenizer.convert_tokens_to_string(token_list)
print(origin_text)
```

```python
random.seed(1)

print(tokenizer.mask_token_id)
print(tokenizer.mask_token)

mask_id = tokenizer.mask_token_id
input_len = len(token['input_ids'])
indices = random.sample(range(input_len), int(input_len * P_MASK))
for idx in indices:
    token['input_ids'][idx] = mask_id
```

```python
token_list = tokenizer.convert_ids_to_tokens(token['input_ids'])
masked_text = tokenizer.convert_tokens_to_string(token_list)
print(masked_text)
```
