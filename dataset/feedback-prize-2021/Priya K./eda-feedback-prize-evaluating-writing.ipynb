## Text Visualization

```python
df['text'] = df['discourse_text'].groupby(df['id']).transform(lambda x: ' '.join(x)) 

df.text.iloc[0]
```

```python
#text length
textlength = df['text'].drop_duplicates().apply(len)
textlength.plot(kind='hist')
plt.title('Essay Length Distribution')
plt.xlabel("Essay Length")
plt.ylabel("Frequency");
```

```python
word_count = df['text'].drop_duplicates().apply(lambda x: len(str(x).split()))
word_count.plot(kind='hist')
plt.title('Word Count Distribution')
plt.xlabel("Word Count")
plt.ylabel("Frequency");
```

```python
colors = {'Lead': 'red','Position': 'green',
          'Claim': 'blue','Evidence': 'yellow',
          'Counterclaim': 'coral','Concluding Statement': 'indigo',
          'Rebuttal': "magenta"}
```

```python
def read_essay(id):
    with open(f"../input/feedback-prize-2021/train/{id}.txt") as f:
        essay = f.read()
    return essay


def visualize(example):
    ents = []
    for i, row in df[df['id'] == example].iterrows():
        ents.append({
                        'start': int(row['discourse_start']), 
                         'end': int(row['discourse_end']), 
                         'label': row['discourse_type']
                    })

    data = read_essay(example)

    doc2 = {
        "text": data,
        "ents": ents,
        "title": example
    }

    options = {"ents": df.discourse_type.unique().tolist(), "colors": colors}
    displacy.render(doc2, style="ent", options=options, manual=True, jupyter=True)
```

```python
visualize("0114B2C51347")
```
