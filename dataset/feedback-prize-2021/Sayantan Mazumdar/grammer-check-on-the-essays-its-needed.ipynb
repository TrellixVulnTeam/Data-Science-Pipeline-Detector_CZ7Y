# Load the essays 

````python
from glob import glob
from tqdm import tqdm
texts=list(map(lambda x:(''.join(open(x).readlines())).replace('/n',''),tqdm(glob("../input/feedback-prize-2021/train/*.txt"))))
````

````python
texts[0]
````

# ```tool.check``` is a function which can give us the number of errors or changes rather than the corrected sentence itself we only take in the first 1000 samples because of faster compute

````python
corrected=list(map(tool.check,tqdm(texts[:1000])))
````

````python
len(corrected[0])
````

````python
import numpy as np
lengths=np.array(list(map(len,corrected)))
````

````python
lengths=lengths/np.array(list(map(lambda x:len(x.split()),texts[:1000])))
````

# Distribution of error percentage

````python
import matplotlib.pyplot as plt
import seaborn as sns
plt.hist(lengths)
````

````python
np.mean(lengths)
````

````python
np.max(lengths)
````

# But what threshold for the error percentage can we really consider as a actually "incorrect" essay I am keeping it was 3 percent of average percentage

````python
import numpy as np
threshold=(1-0.03)*np.mean(lengths)
mask=np.where(lengths>threshold,1,0)
````

````python
print(np.sum(mask)/mask.shape[0])
````

````python
plt.hist(mask)
````
