# Evaluation

投稿された単語は、正解値と予測された単語インデックスの重なり具合で評価されます。

各サンプルについて、与えられたクラスに関するすべての正解値と予測が比較される。<br>
正解値と予測のオーバーラップが >= 0.5 で、予測と正解値のオーバーラップが >= 0.5 の場合、予測はマッチし、真陽性とみなされます。<br>
複数のマッチが存在する場合、最も高いオーバーラップの組を持つマッチが選ばれる。<br>
マッチしていない正解値は偽陰性、マッチしていない予測は偽陽性となります。

例

正解値

```python
# id,class,predictionstring
# 1,Claim,1 2 3 4 5
# 1,Claim,6 7 8
# 1,Claim,21 22 23 24 25
```

予測値

```python
# id,class,predictionstring
# 1,Claim,1 2
# 1,Claim,6 7 8
```

最初の予測は、どちらの正解値とも0.5以上のオーバーラップがなく、偽陽性となるであろう。<br>
2番目の予測は2番目の正解値と完全にオーバーラップし、真陽性となる。<br>
3番目の正解値はマッチングせず、偽陰性となる。

最終的なスコアは、各クラスについてTP/FP/FNを計算し、全クラスにわたるマクロF1スコアを取ることによって得られる。

単語のインデックスはPythonの.split()関数を用いて計算し、結果のリストのインデックスを取る。<br>
2つのオーバーラップは、正解/予測ペアのインデックスの各リストのset()を取り、各セットの長さで割った2つのセットの交点を計算することにより算出されます。
