```python
!pip install textstat

import pandas as pd
import textstat
import matplotlib.pyplot as plt
import os
```

```python
df = pd.read_csv("../input/feedback-prize-2021/train.csv")
df_ss = pd.read_csv("../input/feedback-prize-2021/sample_submission.csv")
```

```python
def read_train_file(currid = "423A1CA112E2", curr_dir = "../input/feedback-prize-2021/train"):
    with open(os.path.join(curr_dir, "{}.txt".format(currid)), "r") as f:
        filetext = f.read()
        
    return filetext
```

```python
scores = []
for i in df["id"]:
    text = read_train_file(i) 
    score = textstat.automated_readability_index(text)
    scores.append(score)
    
scores = [20 if x>20 else x for x in scores]
```

```python
plt.hist(scores, density=False, bins=30)  # density=False would make counts
plt.ylabel('Count')
plt.xlabel('Automated Readability Index');
```
