```python
def get_url_from_gdrive_confirmation(contents):
    url = None
    soup = bs(str(contents))
    downloadForm = soup.select_one("#downloadForm")
    if downloadForm:
        url = downloadForm.attrs.get("action")
    
    if url is None:
        url = get_url_from_gdrive_confirmation_v0(contents)
    
    return url
```

```python
gdown.download.get_url_from_gdrive_confirmation = get_url_from_gdrive_confirmation
```

```python
url = 'https://drive.google.com/drive/folders/1ikKpyt_jVuV26GoUS7x1uQ4mfX-Fr4IK?usp=sharing'

gdown.download_folder.download_folder(url, quiet=False, use_cookies=True, )
```
