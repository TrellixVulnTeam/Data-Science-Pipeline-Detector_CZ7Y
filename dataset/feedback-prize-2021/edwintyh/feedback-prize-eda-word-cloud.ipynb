```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
import re
from wordcloud import WordCloud, STOPWORDS
```

```python
DIR = '../input/feedback-prize-2021'
```

```python
train = pd.read_csv(f'{DIR}/train.csv')
train.head()
```

```python
print (f"Number of Texts: {len(train['id'].unique())}")
print (f"Number of : Dicourses: {len(train['discourse_id'].unique())}")
```

# Target Distribution

```python
fig, ax = plt.subplots()
ax = sns.countplot(data = train, x = 'discourse_type')
ax.set_xticklabels(ax.get_xticklabels(), rotation = 45, ha = 'right')
ax.set_ylabel('Number of Discourse')
ax.set_xlabel('Discourse Type')
plt.show()
```

# Word Count Distribution

```python
train['word_count'] = train['discourse_text'].apply(lambda x: len(x.split()))
discourse_types = train['discourse_type'].unique()
colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown', 'pink']

for idx, discourse_type in enumerate(discourse_types):
    fig, ax = plt.subplots()
    ax = sns.histplot(x = train.loc[train['discourse_type'] == discourse_type, 'word_count'].values, binwidth = 10, stat = 'percent', color = colors[idx], label = discourse_type, alpha = 0.5)
    ax.set_xlim(0, 300)
    plt.legend()
```

# Word Cloud

```python
for discourse_type in discourse_types:

    comment_words = ''
    stopwords = set(STOPWORDS)

    # iterate through the csv file
    for val in train.loc[train['discourse_type'] == discourse_type, 'discourse_text']:

        # typecaste each val to string
        val = str(val)

        # split the value and lowercase
        tokens = [i.lower() for i in re.split('\,|\.|\s', val) if i != '']
        comment_words += " ".join(tokens)+" "
        
        
    wordcloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                stopwords = stopwords,
                min_font_size = 10).generate(comment_words)
    
    print (f'********** {discourse_type} **********')
    # plot the WordCloud image                      
    plt.figure(figsize = (8, 8), facecolor = None)
    plt.imshow(wordcloud)
    plt.axis("off")
    plt.tight_layout(pad = 0)
    plt.show()
```
