# Lets tokenize to see sentiments 

```python
import matplotlib.pyplot as plt
from nltk.tokenize import word_tokenize
s = word_tokenize(x)

plt.hist(s, bins = 500)
plt.show()
```

# lets see Negative sentiments

```python
import string
from wordcloud import WordCloud
wordcloud = WordCloud(background_color="white",width=1600, height=800).generate(' '.join(s))
plt.figure( figsize=(20,10), facecolor='k')
plt.imshow(wordcloud)
```

# most negative comments here relate to morbid assumptions and usage by phones as shown

```python
import seaborn as sns
plt.figure(figsize=(10,10))
ax = sns.countplot(x=df["discourse_type"],  data=df, order = df["discourse_type"].value_counts().index )
for p, label in zip(ax.patches, df["discourse_type"].value_counts()):   
    ax.annotate(label, (p.get_x()+0.25, p.get_height()+0.5))
```
