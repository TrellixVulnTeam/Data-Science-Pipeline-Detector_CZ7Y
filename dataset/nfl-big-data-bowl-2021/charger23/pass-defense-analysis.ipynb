# Pass Defense Analysis

## Introduction 
To understand and evaluate pass defense we will first take a quick look at pass offense characteristics in respect to route success [completion rate] overall and inside the red zone. We will examine team route defense for LAC and then apply these insights as new commentary to Week 15 LAC vs KC matchup.

## Pass Offense Characteristics

First we try to understand more about what we are trying to defend against. We will take a look at route success overall and inside the red zone. See [Table 1] below which includes 13k pass attempts of data from all teams Week 1 thru Week 15.

Some takeaways from [Table 1]:
1. Average pass offense success [completion] rate overall is about 64%. Rate will vary by team.
2. GO route has 37% success rate overall but is the highest TD contributor (156 TD). Classic example of high risk high return. The success rate is so low b/c teams are, at a bare minimum and successful to some extent, defending against this with at least a Cover 1, 1 Free Safety (FS) assisting in the back field.
3. Average red zone success rate reduces to 57%, harder to score in the red zone (w/in 20 yard to touchdown).
4. Approx 70% (449/655) of all TDs comes from inside the red zone - majority of damage is done inside the red zone.
5. CROSS route is quite effective inside the red zone, maintaning 66% success rate inside the zone, same as overall.

In [Table 2] below we will take a look at pass offense characteristics specifically for Kansas City Chiefs (KC), the #1 power offense in 2018.

Some takeaways from [Table 2]:
1. High success rate of 70% overall. High usage of CROSS route, obtained 7 TDs overall.
2. Red zone TD of 29 account for 75% of overall pass TDs, most of the damage happen inside the red zone.
3. What is insane is CROSS route completion rate jumped up by 10% inside the red zone to 79%.
4. Success rate is 69% inside the red zone, over 10% higher than the average 57%.
5. Overall HITCH, FLAT, CROSS are their main offense routes.

KC also has a very good OUT/FLAT route offense, 80% success overall and 80% in the red zone. Every team has different strength and weakness and route usage will vary by team.

## Route Defense

With the various completion rates as benchmarks and noting the various offense route tendencies - we will examine Los Angeles Chargers (LAC) pass defense profile in [Table 3] below.

Main takeways from [Table 3] include:
1. LAC overall completion rate allow is 66% semi on par with the average 64%. Red zone completion rate allowed is 55% slightly better than the average 57%.
2. LAC is fairly effective against the HITCH, CORNER, IN, and POST routes overall and in the red zone. 
3. LAC is decent against the CROSS route at 57% vs avg 66% overall, but allowed 4 for 4 inside the red zone.
3. LAC is more generous with FLAT and SCREEN routes, with 85% to 100% completion rate allowed in the red zone.

The high usage and efficiency with HITCH and CROSS routes that target the mid level (OLB/LB linebacker level) may suggest that there can be potential weakness in cases of zone coverage were linebacker have coverage roles.

Against KC,although LAC is slighly weaker against the short FLAT route, but is above average against HITCH and CROSS. From long route perspective, they are better at POST and weaker against GO routes.
