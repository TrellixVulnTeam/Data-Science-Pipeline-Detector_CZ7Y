### A repository to animate the player tracking within a flask webapp


Apologies for not figuring out the ideal method to share this code via kaggle, but I have a flask webapp developed
to animate each play of this dataset, allowing you to sort through the games, and then the plays within that game.
For each play, you can also shift the time scale forwards and backwards, and selects specific players to see their
information and track line.

The code for this flask app can be found here: https://github.com/hatfieldm48/sports_animation/tree/big_data_bowl_2021

Be sure to use the big_data_bowl_2021 branch. The README should provide all the information you'll need to stand up
the flask webapp locally. If you have any question, please feel free to reach out to me via kaggle or github, and I'll
be happy to provide more information directly, and update the instructions as well.
