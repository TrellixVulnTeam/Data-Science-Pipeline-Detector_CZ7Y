# Evaluating Individual Coverage After the Throw

Scott Fields and George Zhang

## Introduction
On a fundamental level, there are two parts of every passing play in the NFL, before and after the pass is thrown, both of which have distinct goals for individual players. Before the pass is thrown a defender in coverage is either trying to stay with the man they are guarding or trying to cover their zone, but after the ball is released all bets are off and every player on defense has the same goal of either trying to stop the ball before it reaches the intended receiver or tackle the ball carrier after it has arrived. 

We decided to look at coverage after the ball is released and quantify how well individual players contributed to that coverage. We developed DYSAT, or Defensive Yards Saved After the Throw, which aims to create an expected amount of yards gained on any given play based on the tracking data of the main defender at the time of release, the receiver, the ball, and other defenders to judge how much the main defender's actions after the release effected how many yards were gained vs expected. 

## Data
For this research we used the data provided by NFL Big Data Bowl and specifically the following variables: 
* Down
* Yards to Go for 1st Down
* Receiver Tracking Data at Ball Release and Arrival
    * X, Y, Speed, Acceleration, Orientation, Direction
* Defender Tracking Data at Ball Release
    * X, Y, Speed, Acceleration, Orientation, Direction
* Ball Tracking Data at Ball Release and Arrival
    * X, Y, Speed, Acceleration
* Closest 3 Defender Distances at Ball Arrival (not including main defender)

We defined the main defender as the defender closest to the targeted receiver at the time of the ball's release but the research could also be done to evaluate additional relevant defenders on every play.

## Model
The goal of the model was to control for everything outside of the main defender's control and then use that to find patterns of players consistently over or underperforming these expectations. 

The model is not intended to be all inclusive for individual defensive coverage, but instead only grade player's reactions and contest skills once the ball is in the air. For example, if the main defender is running in the opposite direction of the targeted receiver at the time of the ball's release, the expectation would probably be a catch and yards gained, and then if the expection happened the defender won't be penalized very harshly because the mistake was made before the ball was released and has no bearing on the defender's ability to cover after the release.

Or conversely, if a player plays great initial coverage but then gets Mossed and misses a tackle, this model will judge them harshly because we expected them to break up the pass based on how close they were to the receiver. 

For the model we used XGBoost with 10 fold CV to predict the playResult category. The model RMSE of 8.61 yards on the test data and 5.29 yards on all of the data which we think could definitely be improved with more time spent on the model but we also think most of this error can be explained by the model's reluctance to predict gains of 0 yards, the result of which will be DYSAT heavily favoring defenders who can cause incompletions which makes sense.

## Applications
As we said before, DYSAT is not a complete model of defensive performance but we believe when paired with a model evaluating performance before the ball is thrown, DYSAT can help in player evaluation and identifying aspects of players games that need to be worked on that can then be applied in the film room and on the practice field. DYSAT can be filtered to see how player performance changes in certain game situations, different areas of the field, or against certain matchups.
