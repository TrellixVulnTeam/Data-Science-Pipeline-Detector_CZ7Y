First, we identify potential columns for out target/outcome variable based on the data they provide. Here we find that we can visualize outcome using passResult and playResult as they are directly related to the outcome of each individual play. Then we identify the unique values in our chosen outcome variable. Finally, we group our data by relevant outcomes which are in this case domiciled in the passResult column: completed,  incomplete, interception, sacked

```python
#Print unique column values
col_values = df_plays['passResult'].values.ravel()
value = pd.unique(col_values)
print("Unique Values in passResult:", value)
#"Unique Values in passResult:{}".format(value)

col_values = df_plays['playResult'].values.ravel()
value = pd.unique(col_values)
#"Unique Values in playResult:{}".format(value)
print("Unique Values in playResult:", value)




```

Next we try to find out how descriptive downs are and identify who are the most dominant teams across downs. We segment our data by downs 1, 2 and 3 and try to identify relationships that may explain one teams dominance over another. 

```python
#Best performing defenses per down.
down_1st =  df_plays.loc[(df_plays["down"] == 1)]
down_2nd = df_plays.loc[(df_plays["down"] == 2)]
down_3rd = df_plays.loc[(df_plays["down"] == 3)]
```

Next, because we are interested in working backwards from outcome, we group our data by relevant outcomes in our 2 target variables, the first of which is the passResult column with relevant variables: completed,  incomplete, interception, sacked.

```python
#Separate 1st down play data by outcome for 1st down.
#completed plays 1st 
completed_1st = pd.DataFrame(down_1st.loc[(down_1st['passResult'] == "C")])
#incomplete plays 
incomplete_1st = pd.DataFrame(down_1st.loc[(down_1st['passResult'] == "I")])
#intercepted plays
intercepted_1st = pd.DataFrame(down_1st.loc[(down_1st['passResult'] == "IN")])
#sacked plays
sacked_1st = pd.DataFrame(down_1st.loc[(down_1st['passResult'] == "S")])

#extract raw setup of defense setup by number and position of DE, DL and DB. 
completed_1st['defense_str'] = completed_1st.personnelD.str.findall('(\d+)') 
incomplete_1st['defense_str'] = incomplete_1st.personnelD.str.findall('(\d+)')
intercepted_1st['defense_str'] = intercepted_1st.personnelD.str.findall('(\d+)')
sacked_1st['defense_str'] = sacked_1st.personnelD.str.findall('(\d+)')

#pl_team = completed_1st[completed_1st["yardlineSide"] == "KC"]
#Plot plays by favorite offensive dropback type.
team_pass_pl = completed_1st.groupby("typeDropback").count()["defendersInTheBox"].to_frame()
def_drop = team_pass_pl.rename(index=str, columns={"defendersInTheBox" : "num_defenders"})
def_drop.plot(kind='bar')


```

From this visualization we can see that most plays occur and result in the traditional quarterback dropping into the hole play. This means that defenses will face traditional plays most of the time if they are facing a passing play.

Now, we attempt to identify the team on defense by utilizing the yardlineSide column against the team in possession of the ball. If they aren't the same then the yardlineSide team is on defense. 

```python

#Find the team on defense  
plays_team_inc = incomplete_1st[incomplete_1st["yardlineSide"] != incomplete_1st['possessionTeam']]
plays_team_inc.head()

```

Analyzing available data on interceptions. Here we try to plot the amount of yardage won by the defense of each team at first down. We can see that Miami was the most effectivve team at winning yards on the** 1st down.

```python

#Visualize by playResult who successfully defended the most yards for their team on 1st down.
np.exp(plays_team_inc[plays_team_inc['down'] == 1].groupby('yardlineSide')['playResult'].mean()).sort_values().plot(kind='barh',figsize=(12,8))
```

Due to the dynamic nature of football, current score and time play a role in the formation and development of each individual set which affects our outcomes. let us try to visualize this. 

```python
#Scatterplot of gameclock against passresult.
df_plays["gameClock"] = pd.to_datetime(df_plays['gameClock'])
df_plays.plot(kind='scatter',x='gameClock', y='playResult', figsize=(12,20))
```

We find that as expected most plays tend to return between a 0 - 15 point outcome, and most plays are completions.
