Just a few words on a priori. Comming into the season, we have statistics from previous season, such as "Eagles scored, on average, 24 points per game". Their first game of the season updates the expected score by "A priori" x (total number games to be played - game number + 1) + "average score games this season" x (game number - 1).
In each game, we have this adjusted expected score. This adjusts in-game as "Adjusted a priori" x (1 - t) + m, where m is current score, and t the game clock.

Of course, with more statistics, we can utilize things such as arena, weather, home/away et.c.
