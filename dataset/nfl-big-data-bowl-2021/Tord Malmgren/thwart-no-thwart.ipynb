I use datasets games, plays, week1, week2 and week4.

For external data, I checked Espn for the w/l/d.

As the game starts, we initialize a discriminator.

Input to the discriminator are "distance to closest ally", "distance to closest opponent" and "distance to ball". 

For each playId, frameId, nflId, these are added.

Thus, before the tick, we can discriminate thwart/no thwart of each defence.

I.e. we learn the defence of each team as we go.

I did this for four games; PHI-ATL w1, IND-CIN w1 (testing), LA-ARI w2 (biggest difference) and CIN-ATL w4 (closest != 0).

So,

.                   Accuracy

PHI-ATL w1: 18-12, .643-.626

IND-CIN w1: 23-34, .652-.652

LA -ARI w2: 34- 0, .658-.630

CIN-ATL w4: 37-36, .690-.607

which shows it is not too difficult to predict whether a play will advance.
I have not compared this to the null-hypothesis.
