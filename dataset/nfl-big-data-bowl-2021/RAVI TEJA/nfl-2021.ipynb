```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.dates import DateFormatter
```

Reas csv file

```python
game=pd.read_csv("../input/nfl-big-data-bowl-2021/games.csv",parse_dates=['gameDate'])
```

Top five rows to check data types

```python
game.head()
```

```python
game.info()
```

*NO of game played by each team*

```python
plt.figure(figsize=(20,8))
o=game["homeTeamAbbr"].value_counts().index
sns.countplot(data=game,x="homeTeamAbbr",order=o)
plt.xlabel("TEAM")
plt.ylabel("NO OF GAMES")
plt.title("NO OF GAMES PLAYED BY EACH TEAM")
```

INSIGHT:Each team played 8 games, where as team DEN,MIN AND LAC played 7 games each.

In which month most of the game are played.

```python
plt.figure(figsize=(14,8))
game['Month'] = pd.DatetimeIndex(game['gameDate']).month
order1=game["Month"].value_counts().index
sns.countplot(data=game,x="Month",order=order1)
plt.ylabel("NO OF GAMES PLAYED")
plt.xlabel("Month")
plt.title("No of game played per month")
```

INSIGHT:Here we can observe that 12 month most of the games are played.

At what time most of the game are played.

```python
plt.figure(figsize=(14,8))
order2=game["gameTimeEastern"].value_counts().index
sns.countplot(data=game,x="gameTimeEastern",order=order2)
plt.xlabel("Time scale")
plt.ylabel("No of matches")
plt.title("No of matches at particular time")
```

INSIGHT:Most of the game are played in NOON i.e 13:00:00

Per week how many games are played.

```python
plt.figure(figsize=(14,8))
order3=game["week"].value_counts().index
sns.countplot(data=game,x="week",order=order3)
plt.title("NO OF GAME PER WEEK")
```

INSIGHT:
In  most of the week 16 games are played. where as in 9,11,1 only 13 games are played.

                                             **players data**

```python
players=pd.read_csv("../input/nfl-big-data-bowl-2021/players.csv",parse_dates=["birthDate"])
```

```python
players.info()
```

Height measurement conversion from feet to inchs.

```python
f=players["height"].str.split("-",expand=True)
f.columns=["1","2"]
f.loc[(f["2"].notnull()),"1"]=f[f["2"].notnull()]["1"].astype(np.int16)*12+f[f["2"].notnull()]["2"].astype(np.int16)
players['height'] = f['1']
players['height'] = players['height'].astype(np.float32)
```

Height and Weight distrubtion.

```python
x=players["height"]
y=players["weight"]
fig,axes=plt.subplots(1,2)
sns.distplot(x,ax=axes[0],bins=13)
sns.distplot(y,ax=axes[1],bins=13)
plt.show()
```

```python
plt.figure(figsize=(14,8))
order4=players["position"].value_counts().index
sns.countplot(data=players,x="position",order=order4)
plt.title("NO OF PLAYERS PER POSITION")
```

Most of the players are in a WR  position.
Where as at HB,NT and K have very less position members.
