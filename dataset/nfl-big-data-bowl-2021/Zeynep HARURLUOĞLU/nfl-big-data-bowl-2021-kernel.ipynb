```python
traces = [
    go.Histogram(
        x = weeks[mix[i]].value_counts().index, 
        name=mix[i]
    ) for i in range(len(mix))
]

fig = make_subplots(rows=3, cols=3,subplot_titles=(mix))
for i in range(len(traces)):
    fig.append_trace(
        traces[i], 
        (i // 3) + 1, 
        (i % 3) + 1
    )

fig.update_layout(
    template="plotly_white",
    height=1000
)

fig.show()
```
