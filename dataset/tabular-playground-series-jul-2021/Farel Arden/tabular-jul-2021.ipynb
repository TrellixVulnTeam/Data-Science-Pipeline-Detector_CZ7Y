## Separating the Data

```python
x = data_train.iloc[:,1:9].values
y=data_train.iloc[:,9:12].values
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
multilinear = LinearRegression()
multilinear.fit(x_train,y_train)
```

```python
y_pred=multilinear.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import r2_score
accuracy = r2_score(y_test,y_pred)
```

```python
accuracy
```

## Evaluation the Model

<b>MRE</b>

The lower the MSE the better
https://www.youtube.com/watch?v=VRnFon3HYt8

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
print("RMSE value: {:.2f}".format(rmse))
```

```python
print("Train set score: {:.2f}".format(multilinear.score(x_train,y_train)))
```

```python
print("Test set score: {:.2f}".format(multilinear.score(x_test,y_test)))
```
