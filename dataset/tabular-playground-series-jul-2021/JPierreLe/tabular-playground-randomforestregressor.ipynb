# 1. Import libraries

```python
# Import libraries
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

# 2. Data loading

```python
# Load "train.csv", "test.csv" and "submission.csv"
train = pd.read_csv('/kaggle/input/tabular-playground-series-jul-2021/train.csv')
test = pd.read_csv('/kaggle/input/tabular-playground-series-jul-2021/test.csv')
sub = pd.read_csv('../input/tabular-playground-series-jul-2021/sample_submission.csv')
```

```python
# Set features
features1 = ['deg_C', 'relative_humidity', 'absolute_humidity',
       'sensor_1', 'sensor_2', 'sensor_3', 'sensor_4', 'sensor_5']
features2 = ['deg_C', 'relative_humidity', 'absolute_humidity',
       'sensor_1', 'sensor_2', 'sensor_3', 'sensor_4', 'sensor_5']
features3 = ['deg_C', 'relative_humidity', 'absolute_humidity',
       'sensor_1', 'sensor_2', 'sensor_3', 'sensor_4', 'sensor_5']
# Set inputs
X1 = train[features1]        # input for training
X_test1 = test[features1]    # input for testing
X2 = train[features2]        # input for training
X_test2 = test[features1]    # input for testing
X3 = train[features3]        # input for training
X_test3 = test[features3]    # input for testing

# Set outputs
Y1 = train["target_carbon_monoxide"]    # output for training
Y2 = train["target_benzene"]            # output for training
Y3 = train["target_nitrogen_oxides"]    # output for training
```

# 3. Hyper parameter

```python
# Set hyper parameter for selected model(Random Forest)
clf1 = RandomForestRegressor(n_estimators=100, max_depth=5, random_state=1)
clf2 = RandomForestRegressor(n_estimators=100, max_depth=5, random_state=1)
clf3 = RandomForestRegressor(n_estimators=100, max_depth=5, random_state=1)
```

```python
# Set hyper parameter for k-fold cross validation
k_fold = KFold(n_splits=10, shuffle=True, random_state=0)
```

# 4. Cross validaion

```python
# K-fold cross validation to predict Y1
score1 = np.sqrt(-1*cross_val_score(clf1, X1, Y1, cv=k_fold, scoring="neg_mean_squared_log_error"))
print(np.average(score1))
```

```python
# K-fold cross validation to predict Y2
score2 = np.sqrt(-1*cross_val_score(clf2, X2, Y2, cv=k_fold, scoring="neg_mean_squared_log_error"))
print(np.average(score2))
```

```python
# K-fold cross validation to predict Y3
score3 = np.sqrt(-1*cross_val_score(clf3, X3, Y3, cv=k_fold, scoring="neg_mean_squared_log_error"))
print(np.average(score3))
```

```python
# Show MCRMSLE (Mean Columnwise Root Mean Squared Logarithmic Error)
print(np.average(score1+score2+score3)/3)
```

# 5. Training

```python
clf1.fit(X1, Y1)
```

```python
clf2.fit(X2, Y2)
```

```python
clf3.fit(X3, Y3)
```

# 6. Prediction

```python
pred1 = clf1.predict(X_test1)
```

```python
pred2 = clf2.predict(X_test2)
```

```python
pred3 = clf3.predict(X_test3)
```

# 7. Submission

```python
sub.target_carbon_monoxide = pred1
sub.target_benzene = pred2
sub.target_nitrogen_oxides = pred3
sub.to_csv('addfeatures_submission.csv',index=False)
sub
```
