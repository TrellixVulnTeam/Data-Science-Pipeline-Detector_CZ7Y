```python
import numpy as np
import math
import pandas as pd
import warnings
import seaborn as sns
import matplotlib.pyplot as plt

warnings.filterwarnings("ignore")
sns.set_style("darkgrid")
```

```python
train = pd.read_csv("/kaggle/input/tabular-playground-series-jul-2021/train.csv")
test  = pd.read_csv("/kaggle/input/tabular-playground-series-jul-2021/test.csv")
```

## Plot a pair grid for marginal probablity analysis and relationship analysis for 
* deg_c
* relative humidity
* absolute humidity

* > it looks they follow gaussian distribution, with simply ignorable range of outliers


```python
Grid = sns.PairGrid(train,corner=True,diag_sharey=False,vars=train.columns[1:4],layout_pad=1,size=5) #corner helps to remove redundant mirror plots
Grid.map(sns.scatterplot)
Grid.map_offdiag(sns.scatterplot,alpha=0.5,s=30,ec="red",color=".6")
Grid.map_diag(sns.histplot,ec="red")
```

### Findings on relationship plot 2
* from the below graph we can understand the target vars are skewed, so we use log transform to scale it
* the independent variables suits for a typical linear regression based on the visualization 

```python
grid2 = sns.PairGrid(train,corner=True,diag_sharey=False,vars=train.columns[5:13])
grid2.map_diag(sns.histplot)
grid2.map_offdiag(sns.scatterplot,ec="red",alpha=0.6)
```
