In order to get quick insights to a dataset in few minutes, I always start by doing a data profiling and then I do other visualizations to push the analysis further more.

You can find the docs of the librar in the link below: https://pandas-profiling.github.io/pandas-profiling/docs/master/rtd/index.html

```python
train = pd.read_csv('/kaggle/input/tabular-playground-series-jul-2021/train.csv')
test = pd.read_csv('/kaggle/input/tabular-playground-series-jul-2021/test.csv')
train.head(3)
```

```python
print('size of the train set:', train.shape)
print('size of the test set:', test.shape)
```

```python
target_carbon = train['target_carbon_monoxide']
target_benzene = train['target_benzene']
target_nitrogen = train['target_nitrogen_oxides']
targets = pd.concat([target_carbon, target_benzene, target_nitrogen] , axis=1)
X_train = train.drop(targets.columns, axis=1)
```

```python
profile = ProfileReport(train, title="Pandas Profiling Report", explorative=True)
profile.to_notebook_iframe()
```
