```python
sample_submission= pd.read_csv('/kaggle/input/tabular-playground-series-jul-2021/sample_submission.csv')
sample_submission.shape
```

```python
model_1= pd.read_csv('/kaggle/input/tps-eda-lstm-pseudolabels-interpolate/Submission.csv')
model_2=pd.read_csv('/kaggle/input/k/alexryzhkov/k/alexryzhkov/tps-lightautoml-baseline-with-pseudolabels/lightautoml_with_pseudolabelling_kernel_version_16.csv')
                     
```

```python
total_data= pd.concat([model_1,model_2]).groupby(level=0).mean()
```
