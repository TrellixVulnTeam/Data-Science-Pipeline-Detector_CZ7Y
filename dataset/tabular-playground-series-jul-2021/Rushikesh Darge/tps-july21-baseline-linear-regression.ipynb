```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

In this competition you are predicting the values of air pollution measurements over time, based on basic weather information (temperature and humidity) and the input values of 5 sensors.

The three target values to you to predict are: `target_carbon_monoxide, target_benzene, and target_nitrogen_oxides`

```python
train = pd.read_csv('../input/tabular-playground-series-jul-2021/train.csv')
test = pd.read_csv('../input/tabular-playground-series-jul-2021/test.csv')
submission = pd.read_csv('../input/tabular-playground-series-jul-2021/sample_submission.csv')
train.head()
```

```python
test.head()
```

```python
submission.head()
```

```python
train.columns
```

```python
cols = test.columns[1:]
cols
```

```python
X = train[cols]
target1 = train['target_carbon_monoxide']
target2 = train['target_benzene']
target3 = train['target_nitrogen_oxides']
```

```python
from sklearn.linear_model import LinearRegression
reg1 = LinearRegression().fit(X, target1)
reg1.score(X, target1)
```

```python
reg2 = LinearRegression().fit(X, target2)
reg2.score(X, target2)
```

```python
reg3 = LinearRegression().fit(X, target3)
reg3.score(X, target3)
```

```python
submission.head()
```

```python
test.head(2)
```

```python
submission['date_time'] = test['date_time']
```

```python
submission['target_carbon_monoxide'] = reg1.predict(test[cols])
submission['target_benzene'] = reg2.predict(test[cols])
submission['target_nitrogen_oxides'] = reg3.predict(test[cols])
```
