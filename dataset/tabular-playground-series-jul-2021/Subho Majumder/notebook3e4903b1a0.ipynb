```python
import numpy as np
import pandas as pd
import sklearn as sk
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

from sklearn.svm import SVR
from sklearn.naive_bayes import GaussianNB


from sklearn.model_selection import train_test_split
from sklearn.linear_model import ElasticNet,MultiTaskLassoCV,RidgeCV,LinearRegression
from sklearn.tree import DecisionTreeRegressor

from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.experimental import enable_hist_gradient_boosting
from sklearn.ensemble import HistGradientBoostingRegressor

from sklearn.ensemble import VotingRegressor
from sklearn.ensemble import StackingRegressor

from sklearn.neighbors import KNeighborsRegressor
```

```python
data = pd.read_csv("../input/tabular-playground-series-jul-2021/train.csv")
data.head()
```

```python
x=data.iloc[:,1:9]
y=data.iloc[:,9:]
x=x.to_numpy()
y=y.to_numpy()
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
xtrain.shape
#xtest.shape
```

```python
dic={}
```

```python
enet=ElasticNet()
enet.fit(xtrain,ytrain)
ypred=enet.predict(xtest)
err1=mean_squared_error(ypred[:,0],ytest[:,0])
err2=mean_squared_error(ypred[:,1],ytest[:,1])
err3=mean_squared_error(ypred[:,2],ytest[:,2])
err=mean_squared_error(ypred,ytest)
print(err,err1,err2,err3)
dic['elasticnet']=[err,err1,err2,err3]
```

```python
dtr=DecisionTreeRegressor()
dtr.fit(xtrain,ytrain)
ypred=dtr.predict(xtest)
err1=mean_squared_error(ypred[:,0],ytest[:,0])
err2=mean_squared_error(ypred[:,1],ytest[:,1])
err3=mean_squared_error(ypred[:,2],ytest[:,2])
err=mean_squared_error(ypred,ytest)
print(err,err1,err2,err3)
dic['decision tree']=[err,err1,err2,err3]
```

```python
lasso=MultiTaskLassoCV()
lasso.fit(xtrain,ytrain)
ypred=lasso.predict(xtest)
err1=mean_squared_error(ypred[:,0],ytest[:,0])
err2=mean_squared_error(ypred[:,1],ytest[:,1])
err3=mean_squared_error(ypred[:,2],ytest[:,2])
err=mean_squared_error(ypred,ytest)
print(err,err1,err2,err3)
dic['lasso']=[err,err1,err2,err3]
```

```python
ridge=RidgeCV()
ridge.fit(xtrain,ytrain)
ypred=ridge.predict(xtest)
err1=mean_squared_error(ypred[:,0],ytest[:,0])
err2=mean_squared_error(ypred[:,1],ytest[:,1])
err3=mean_squared_error(ypred[:,2],ytest[:,2])
err=mean_squared_error(ypred,ytest)
print(err,err1,err2,err3)
dic['ridge']=[err,err1,err2,err3]
```

```python
for i in range(10,250,10):
    rf=RandomForestRegressor(n_estimators=i,random_state=7)
    rf.fit(xtrain,ytrain)
    ypred=rf.predict(xtest)
    err1=mean_squared_error(ypred[:,0],ytest[:,0])
    err2=mean_squared_error(ypred[:,1],ytest[:,1])
    err3=mean_squared_error(ypred[:,2],ytest[:,2])
    err=mean_squared_error(ypred,ytest)
    print(i,err,err1,err2,err3)
```

```python
rf=RandomForestRegressor(n_estimators=140,random_state=7)
rf.fit(xtrain,ytrain)
ypred=rf.predict(xtest)
err1=mean_squared_error(ypred[:,0],ytest[:,0])
err2=mean_squared_error(ypred[:,1],ytest[:,1])
err3=mean_squared_error(ypred[:,2],ytest[:,2])
err=mean_squared_error(ypred,ytest)
print(err,err1,err2,err3)
dic['rf']=[err,err1,err2,err3]
```

```python
xt=ExtraTreesRegressor()
xt.fit(xtrain,ytrain)
ypred=xt.predict(xtest)
err1=mean_squared_error(ypred[:,0],ytest[:,0])
err2=mean_squared_error(ypred[:,1],ytest[:,1])
err3=mean_squared_error(ypred[:,2],ytest[:,2])
err=mean_squared_error(ypred,ytest)
print(err,err1,err2,err3)
dic['ExtraTreesRegressor']=[err,err1,err2,err3]
```

```python
knn=KNeighborsRegressor()
knn.fit(xtrain,ytrain)
ypred=knn.predict(xtest)
err1=mean_squared_error(ypred[:,0],ytest[:,0])
err2=mean_squared_error(ypred[:,1],ytest[:,1])
err3=mean_squared_error(ypred[:,2],ytest[:,2])
err=mean_squared_error(ypred,ytest)
print(err,err1,err2,err3)
dic['knn']=[err,err1,err2,err3]
```
