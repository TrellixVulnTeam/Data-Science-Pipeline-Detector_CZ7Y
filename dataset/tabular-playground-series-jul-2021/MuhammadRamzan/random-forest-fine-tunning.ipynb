```python
from sklearn.ensemble import RandomForestRegressor 
```

```python
train = pd.read_csv('/kaggle/input/tabular-playground-series-jul-2021/train.csv')
test = pd.read_csv('/kaggle/input/tabular-playground-series-jul-2021/test.csv')
submission = pd.read_csv('/kaggle/input/tabular-playground-series-jul-2021/sample_submission.csv')
```

```python
columns = test.columns[1:]
columns
```

```python
X_train = train[columns]
X_test = test[columns]
```

```python
target_1 = train['target_carbon_monoxide'].values.reshape(-1,1)
target_2 = train['target_benzene'].values.reshape(-1,1)
target_3 = train['target_nitrogen_oxides'].values.reshape(-1,1)
```

```python
rf_1 = RandomForestRegressor(max_depth=100, max_features='sqrt', min_samples_leaf=3,
                           min_samples_split=12, n_estimators=600,bootstrap=True)
rf_1.fit(X_train, target_1)
submission['target_carbon_monoxide'] = rf_1.predict(X_test)

rf_2 = RandomForestRegressor(max_depth=8, max_features='auto', min_samples_leaf=1,
                           min_samples_split=6, n_estimators=1600,bootstrap=True)
rf_2.fit(X_train, target_2)
submission['target_benzene'] = rf_2.predict(X_test)

rf_3 = RandomForestRegressor(max_depth=40, max_features='sqrt', min_samples_leaf=5,
                           min_samples_split=10, n_estimators=800,bootstrap=True,)
rf_3.fit(X_train, target_3)
submission['target_nitrogen_oxides'] = rf_3.predict(X_test)

submission.head()
```

```python
submission.to_csv('submission_rf_hyper_all.csv', index=False)
```
