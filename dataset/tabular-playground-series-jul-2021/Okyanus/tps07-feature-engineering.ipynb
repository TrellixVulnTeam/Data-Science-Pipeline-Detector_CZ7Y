# Load data

```python
import pandas as pd

train_df = pd.read_csv("../input/tabular-playground-series-jul-2021/train.csv")
test_df = pd.read_csv("../input/tabular-playground-series-jul-2021/test.csv")
```

# Celsius to Fahrenheit

```python
def Celsius_to_F(celsius):
    return (celsius * 1.8) + 32
```

```python
Celsius_to_F(32)
```

```python
train_df["deg_F"] = train_df["deg_C"].map(Celsius_to_F)
test_df["deg_F"] = test_df["deg_C"].map(Celsius_to_F)
```

```python
train_df[["deg_C", "deg_F"]]
```

# Celsius to Kelvin

```python
def Celsius_to_K(celsius):
    return celsius + 273.15
```

```python
train_df["deg_K"] = train_df["deg_C"].map(Celsius_to_K)
test_df["deg_K"] = test_df["deg_C"].map(Celsius_to_K)
```

```python
train_df[["deg_C", "deg_F", "deg_K"]]
```

# Calculate due_point_temperature

```python
def dpt(observation):
    return observation["deg_C"] *((100 - observation["relative_humidity"])/5)
```

```python
train_df["due_point_temperature"] = train_df.apply(dpt, axis=1)
test_df["due_point_temperature"] = test_df.apply(dpt, axis=1)
```

# Saving feature-engineered dataset to CSV

```python
train_df.to_csv("train.csv", index=False)
test_df.to_csv("test.csv", index=False)
```
