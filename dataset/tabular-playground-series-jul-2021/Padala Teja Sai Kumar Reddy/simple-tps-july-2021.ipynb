```python
sensordata=df[['date_time','deg_C','relative_humidity','absolute_humidity','sensor_1','sensor_2','sensor_3','sensor_4','sensor_5']]
targets=df[['target_carbon_monoxide','target_benzene','target_nitrogen_oxides']]
```

```python
sensordata['date_time'] = pd.to_datetime(sensordata['date_time'])
sensordata['date_time'] = sensordata['date_time'].apply(lambda x: x.value)
```

```python
from sklearn import  linear_model
model_enet = linear_model.ElasticNet(alpha = 0.01)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(sensordata,targets, train_size = 0.8, 
                                                    test_size = 0.2, random_state = 100)
```

```python
model_enet.fit(X_train, y_train)
```

```python
y_test_pred= model_enet.predict(X_test)
X_test_pred= model_enet.predict(X_train)
```

```python
from sklearn.metrics import r2_score

r_squared = r2_score(y_test, y_test_pred)
r_squared
```

```python
r_squared = r2_score(y_train,X_test_pred)
r_squared
```
