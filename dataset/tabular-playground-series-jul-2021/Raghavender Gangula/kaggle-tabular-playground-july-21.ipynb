```python
train_file_path = '../input/tabular-playground-series-jul-2021/train.csv'
test_file_path = '../input/tabular-playground-series-jul-2021/test.csv'

train_data = pd.read_csv(train_file_path)
test_data = pd.read_csv(test_file_path)

# print(train_data.info(),'\n')
```

```python
train_data.corr()
```

```python
y_carbon_monoxide = train_data.target_carbon_monoxide
y_benzene = train_data.target_benzene
y_nitrogen_oxides = train_data.target_nitrogen_oxides

X = train_data.drop(['date_time' ,'target_carbon_monoxide', 'target_benzene', 'target_nitrogen_oxides'], axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_valid, y_CO_train, y_CO_valid, y_Ben_train, y_Ben_valid, y_NO_train, y_NO_valid = train_test_split(
    X, y_carbon_monoxide, y_benzene, y_nitrogen_oxides, test_size=0.2, train_size=0.8, random_state=0)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor

my_pipeline1 = Pipeline(steps = [
    ('scaler', StandardScaler()),
    ('model', RandomForestRegressor(random_state=0))
])

my_pipeline2 = Pipeline(steps = [
    ('scaler', StandardScaler()),
    ('model', RandomForestRegressor(random_state=0))
])

my_pipeline3 = Pipeline(steps = [
    ('scaler', StandardScaler()),
    ('model', RandomForestRegressor(random_state=0))
])
# def my_model(X_train, y_train):
#     scaler = StandardScaler()
#     scaler.fit_transform(X_train,y_train)
#     model = RandomForestRegressor(random_state=0)
#     model.fit(X_train, y_train)
#     return model
```

```python
my_pipeline1.fit(X_train, y_CO_train)
preds_CO = my_pipeline1.predict(X_valid)

my_pipeline2.fit(X_train, y_Ben_train)
preds_Ben = my_pipeline2.predict(X_valid)

my_pipeline3.fit(X_train, y_NO_train)
preds_NO = my_pipeline3.predict(X_valid)

# model1 = my_model(X_train, y_CO_train)
# model2 = my_model(X_train, y_Ben_train)
# model3 = my_model(X_train, y_NO_train)

# preds_CO = model1.predict(X_valid)
# preds_Ben = model2.predict(X_valid)
# preds_NO = model3.predict(X_valid)
```

```python
from sklearn.metrics import mean_squared_log_error
msle_CO = mean_squared_log_error(y_CO_valid, preds_CO)
msle_Ben = mean_squared_log_error(y_Ben_valid, preds_Ben)
msle_NO = mean_squared_log_error(y_NO_valid, preds_NO)
print(msle_CO, msle_Ben, msle_NO)
```

```python
np.mean([msle_CO, msle_Ben, msle_NO])
```

```python
X_test = test_data.drop('date_time', axis=1).copy()
predictions_carbon_monoxide = my_pipeline1.predict(X_test)
predictions_benzene = my_pipeline2.predict(X_test)
predictions_nitrogen_oxides = my_pipeline3.predict(X_test)
```

```python
output = pd.DataFrame({'date_time':test_data.date_time, 'target_carbon_monoxide':predictions_carbon_monoxide, 'target_benzene':predictions_benzene, 'target_nitrogen_oxides':predictions_nitrogen_oxides})
output.to_csv('submission.csv', index = False)
```
