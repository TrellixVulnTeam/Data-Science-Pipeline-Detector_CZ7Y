```python
columns = test.columns[1:]
columns
```

```python
X = train[columns].values
X_test = test[columns].values
target_1 = train['target_carbon_monoxide'].values.reshape(-1,1)
target_2 = train['target_benzene'].values.reshape(-1,1)
target_3 = train['target_nitrogen_oxides'].values.reshape(-1,1)
```

```python
from sklearn.linear_model import Ridge
```

```python
rr = Ridge()
rr.fit(X, target_1)
sample_submission['target_carbon_monoxide'] = rr.predict(X_test)
sample_submission['target_carbon_monoxide'] = np.power(sample_submission['target_carbon_monoxide'], 3)

rr = Ridge()
rr.fit(X, target_2)
sample_submission['target_benzene'] = rr.predict(X_test)
sample_submission['target_benzene'] = np.power(sample_submission['target_benzene'], 3)

rr = Ridge()
rr.fit(X, target_3)
sample_submission['target_nitrogen_oxides'] = rr.predict(X_test)
sample_submission['target_nitrogen_oxides'] = np.power(sample_submission['target_nitrogen_oxides'], 3)

sample_submission.head()
```

```python
sample_submission.to_csv('submission_rr_0.csv', index=False)
```

```python
from catboost import CatBoostRegressor
# Initialize data

train_data = train.drop(['target_carbon_monoxide','date_time', 'target_benzene', 'target_nitrogen_oxides' ], axis = 1)

eval_data = test.drop(['date_time'], axis = 1)

train_labels = train['target_carbon_monoxide']
# Initialize CatBoostRegressor
model = CatBoostRegressor(iterations=500,
                          learning_rate=0.15,
                          depth=2)
# Fit model
model.fit(train_data, train_labels)
# Get predictions
preds = model.predict(eval_data)
```

```python
sample_sub1['target_carbon_monoxide'] = preds
```
