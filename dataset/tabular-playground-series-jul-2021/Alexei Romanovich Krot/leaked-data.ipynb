```python
target_carbon_monoxide = data_leaked_f['CO(GT)']
target_benzene = data_leaked_f['C6H6(GT)']
target_nitrogen_oxides = data_leaked_f['NOx(GT)']
```

```python
target_carbon_monoxide[0] = 2.05
target_nitrogen_oxides[0] = 318.89
n = -1
for i in target_carbon_monoxide:
    n += 1
    if i == -200:
        target_carbon_monoxide[n] = (target_carbon_monoxide[n - 1] + target_carbon_monoxide[n + 1]) / 2
nn = -1
for ii in target_benzene:
    nn += 1
    if ii == -200:
        target_benzene[nn] = (target_benzene[nn - 1] + target_benzene[nn + 1]) / 2
nnn = -1
for iii in target_nitrogen_oxides:
    nnn += 1
    if iii == -200:
        target_nitrogen_oxides[nnn] = (target_nitrogen_oxides[nnn - 1] + target_nitrogen_oxides[nnn + 1]) / 2
```

```python
test
```

```python
output = pd.DataFrame({'date_time': pd.read_csv('/kaggle/input/tabular-playground-series-jul-2021/test.csv').date_time,
                       'target_carbon_monoxide': target_carbon_monoxide, 
                       'target_benzene': target_benzene, 
                       'target_nitrogen_oxides': target_nitrogen_oxides})

output.to_csv('submission.csv', index = False)
output
```
