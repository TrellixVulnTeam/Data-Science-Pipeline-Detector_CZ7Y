```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

train = pd.read_csv('../input/tabular-playground-series-jul-2021/train.csv')
train['date_time'] = pd.to_datetime(train['date_time'])
train.head()
```

### This TPS - Jul 's data is time series data. So, We can find appropriate date size !
### First I try divide 1 Day

```python
train = train.set_index('date_time').resample('1D').mean()
fig = plt.figure(figsize=(20,10))
plt.plot(train.index,train['target_carbon_monoxide'])
plt.plot(train.index,train['sensor_1'])
plt.show()
```

### Before Visualization, We need to Normalization

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
train = pd.DataFrame(scaler.fit_transform(train),columns = train.columns,index = train.index)

fig = plt.figure(figsize=(20,10))
plt.plot(train.index,train['target_carbon_monoxide'])
plt.plot(train.index,train['sensor_1'])
plt.show()
```

### I think that it's too hectic to look at the chart.
### So, I want to divide 5 Days

```python
train = pd.read_csv('../input/tabular-playground-series-jul-2021/train.csv')
train['date_time'] = pd.to_datetime(train['date_time'])
train = train.set_index('date_time').resample('4D').mean()
train = pd.DataFrame(scaler.fit_transform(train),columns = train.columns,index = train.index)

fig = plt.figure(figsize=(20,10))
plt.plot(train.index,train['target_carbon_monoxide'])
plt.plot(train.index,train['sensor_1'])
plt.show()
```

### It's better than previous chart. 
### Now, I want to distinguish two charts, So I will change sensor_1 's data lineplot to barplot

```python
fig = plt.figure(figsize=(20,10))
plt.plot(train.index,train['target_carbon_monoxide'])
plt.bar(x= train.index, height=train['sensor_1'])
plt.show()
```

### It's too hectic to look at the chart.
### Add 0.5 value to result value : target_carbon_monoxide

```python
fig = plt.figure(figsize=(20,10))
plt.plot(train.index,train['target_carbon_monoxide']+0.5)
plt.bar(x= train.index, height=train['sensor_1'],width=1)
plt.show()
```

### Control the barplot's opacity, color and linewidth ! ( I change add 0.5 to add 0.3 ! )

```python
fig = plt.figure(figsize=(20,10))
plt.plot(train.index,train['target_carbon_monoxide']+0.3,lw=3)
plt.bar(x= train.index, height=train['sensor_1'],width=1,alpha=0.3,color='red')
plt.show()
```

### Now, We can feel some connection

```python
fig = plt.figure(figsize=(20,10))
plt.plot(train.index,(train['target_carbon_monoxide']-0.5),lw=3)
plt.bar(x= train.index, height=(train['sensor_1']-0.5),width=2,alpha=0.3,color='red')
plt.show()
```

### How about this chart ? I think it's better than previous chart that knowing connection with two things
### Now, Decorate! Decorating is essential to attract people's visual interest.

```python
import matplotlib.font_manager as fm
font = fm.FontProperties(fname='../input/staatfont/Staatliches-Regular.ttf')
```

```python
fig = plt.figure(figsize=(20,10))
plt.plot(train.index,(train['target_carbon_monoxide']-0.5)*1.5,lw=2.5, color='coral')
plt.bar(x= train.index, height=train['sensor_1']-0.5,width=2,alpha=0.3,color='lightcoral')

plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)

plt.xticks(font=font,fontsize=20)
plt.yticks(font=font,fontsize=20)

plt.title('The relationship with result and Sensor No_1', font=font, fontsize=40)

plt.show()
```
