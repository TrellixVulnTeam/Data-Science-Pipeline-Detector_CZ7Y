```python
from sklearn.multioutput import MultiOutputRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
measurements.fillna(measurements.median(), inplace=True)
X = measurements[['deg_C', 'relative_humidity', 'absolute_humidity', 'sensor_1', 'sensor_2', 'sensor_3', 'sensor_4', 'sensor_5']]
y = measurements[['target_carbon_monoxide', 'target_benzene', 'target_nitrogen_oxides']]
```

```python
clf  = MultiOutputRegressor(GradientBoostingRegressor())
clf.fit(X, y)
```

```python
X_test = [['deg_C', 'relative_humidity', 'absolute_humidity', 'sensor_1', 'sensor_2', 'sensor_3', 'sensor_4', 'sensor_5']];
prediction = clf.predict(X_test)
```

```python
submission = pd.read_csv('../input/tabular-playground-series-jul-2021/sample_submission.csv', index_col='date_time')
submission.iloc[:, :] = prediction
submission.to_csv("submission.csv")
```
