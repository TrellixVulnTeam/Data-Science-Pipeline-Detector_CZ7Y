```python
import pandas as pd
```

```python
# weights supplied by the overfitting fairy...
model_paths = [
    (.25, '../input/tps-07-turtles-on-turtles/submission_gam_adjust.csv'),
    (.15, '../input/tps07-jul-2021-pycaret-with-pseudo-labels/submission.csv'),
    (.15, '../input/tps-07-gam-baseline-trend-shuffled-months/submission_gam_no_trend2.csv'),
    (.30, '../input/catboost-14feature-cross-validation-fe/submission.csv'),
    (.10, '../input/tps-lightautoml-baseline-with-pseudolabels/lightautoml_with_pseudolabelling_kernel_version_12.csv'),
    (.05, '../input/2-tps-jul-21-smart-ensembling/submission.csv')]

targets = ['target_carbon_monoxide','target_benzene','target_nitrogen_oxides']

submissions = []
for weight, path in model_paths:
    df = pd.read_csv(path)
    df[targets] = df[targets] * weight
    submissions.append(df)

df_combined = pd.concat(submissions).groupby(['date_time']).sum().reset_index()
```

```python
# completely untrustworthy...
df_combined.to_csv('submission_overfit.csv', index=False)
```
