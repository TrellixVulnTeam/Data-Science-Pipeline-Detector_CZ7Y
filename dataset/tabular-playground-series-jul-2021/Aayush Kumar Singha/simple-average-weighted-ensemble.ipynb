```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
s1 = pd.read_csv("../input/tpsjuly2021bestsubmissions/1_submission_pycaret.csv")
s2 = pd.read_csv("../input/tpsjuly2021bestsubmissions/2_submission_gam.csv")
s3 = pd.read_csv("../input/tpsjuly2021bestsubmissions/3_submission_catboost.csv")
s4 = pd.read_csv("../input/tpsjuly2021bestsubmissions/4_sumbission_autogluon.csv")
```

```python
s = pd.concat([s1, s2, s3, s4]).groupby(level=0).mean()
```

```python
s['date_time'] = s4['date_time']
```

```python
s = s[['date_time', 'target_carbon_monoxide', 'target_benzene', 'target_nitrogen_oxides']]
```

```python
s.to_csv('submission_average_weighted_blend.csv', index=False)
```
