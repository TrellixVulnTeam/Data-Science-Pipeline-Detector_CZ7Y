```python
import xgboost as xgb
import numpy as np
import pandas as pd
import random
import optuna
from sklearn.model_selection import KFold, train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
```

```python
train = pd.read_csv("../input/tabular-playground-series-jul-2021/train.csv")
test = pd.read_csv("../input/tabular-playground-series-jul-2021/test.csv")
```

```python
train.head()
```

```python
print(train.columns.to_list())
```

```python
target_name=['target_carbon_monoxide','target_benzene','target_nitrogen_oxides']
train_name=['deg_C','relative_humidity','absolute_humidity','sensor_1','sensor_2',
            'sensor_3','sensor_4','sensor_5',]
```

```python
target = train[target_name]
target0 = train['target_carbon_monoxide']
data = train[train_name]
```

```python
columns=data.columns.to_list()
print(columns)
```

```python
def objective(trial,data=data,target=target0):
    
    train_x, test_x, train_y, test_y = train_test_split(data, target, test_size=0.2,random_state=42)
    param = {

        'lambda': trial.suggest_uniform('lambda',5.0,6.0),
        'alpha': trial.suggest_uniform('alpha',2.0,2.4),
        'colsample_bytree': trial.suggest_uniform('colsample_bytree', 0.3,0.5),
        'subsample': trial.suggest_uniform('subsample', 0.5,0.6),
        'learning_rate': trial.suggest_uniform('learning_rate', 0.04,0.06),
        'n_estimators': trial.suggest_int('n_estimators', 100,2000),
        'max_depth': trial.suggest_int('max_depth', 15,20),
        'random_state': trial.suggest_int('random_state', 1000,2000),
        'min_child_weight': trial.suggest_int('min_child_weight', 20,60),
        
        'use_label_encoder': trial.suggest_categorical('use_label_encoder',['False']),
        'objective': trial.suggest_categorical('objective',['reg:tweedie']), 
        'tree_method': trial.suggest_categorical('tree_method',['gpu_hist']),  #'gpu_hist','hist'       
    }
    model = xgb.XGBRegressor(**param)      ###
    model.fit(train_x,train_y,eval_set=[(test_x,test_y)],early_stopping_rounds=100,verbose=False)
    preds = model.predict(test_x)
    rmse = mean_squared_error(test_y, preds,squared=False)
    
    return rmse
```

#### Tuning manually repeated and the range of values narrowed.

```python
study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=8)
print('Number of finished trials:', len(study.trials))
print('Best trial:', study.best_trial.params)
```

```python
Best_trial=study.best_trial.params
print(Best_trial)
```

```python
sample = pd.read_csv("../input/tabular-playground-series-jul-2021/sample_submission.csv")
print(sample.shape)
sample
```

```python
preds = np.zeros((sample.shape[0],sample.shape[1]-1))
kf = KFold(n_splits=5,random_state=48,shuffle=True)
for i in range(3):
    for trn_idx, test_idx in kf.split(train[columns],target[target_name[i]]):
        X_tr,X_val=train[columns].iloc[trn_idx],train[columns].iloc[test_idx]
        y_tr,y_val=target.iloc[trn_idx,i],target.iloc[test_idx,i]     ######
        model = xgb.XGBRegressor(**Best_trial)
        model.fit(X_tr,y_tr,eval_set=[(X_val,y_val)],early_stopping_rounds=100,verbose=False)
        preds[:,i]+=model.predict(test[columns])/kf.n_splits    ######
        rmse=mean_squared_error(y_val, model.predict(X_val),squared=False)
        print(rmse)
```

```python
subm = sample
subm[target_name] = pd.DataFrame(preds)
subm.to_csv('submission.csv',index=False)
subm
```
