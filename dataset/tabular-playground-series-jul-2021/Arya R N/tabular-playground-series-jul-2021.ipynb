```python
features = ["deg_C","relative_humidity","absolute_humidity","sensor_1","sensor_2","sensor_3","sensor_4","sensor_5"]
X = train_data[features]
y1 = train_data['target_carbon_monoxide']
y2 = train_data['target_benzene']
y3 = train_data['target_nitrogen_oxides']
```

```python
from sklearn.model_selection import train_test_split
train_X,val_X,train_y1,val_y1 = train_test_split(X,y1)
train_X,val_X,train_y2,val_y2 = train_test_split(X,y2)
train_X,val_X,train_y3,val_y3 = train_test_split(X,y3)
```

```python
from sklearn.neighbors import KNeighborsRegressor

prediction1 = []
prediction2 = []
prediction3 = []
for i in range(5,20):
    model1 = KNeighborsRegressor(i)
    model2 = KNeighborsRegressor(i)
    model3 = KNeighborsRegressor(i)
    model1.fit(train_X,train_y1)
    model2.fit(train_X,train_y2)
    model3.fit(train_X,train_y3)
    val_predictions1 = model1.predict(val_X)
    val_predictions2 = model2.predict(val_X)
    val_predictions3 = model3.predict(val_X)
    prediction1.append(val_predictions1)
    prediction2.append(val_predictions2)
    prediction3.append(val_predictions3)
```

```python
from sklearn.metrics import mean_absolute_error
mae1 = []
mae2 = []
mae3 = []
for i in range(len(prediction1)):
    mae1.append(mean_absolute_error(val_y1,prediction1[i]))
    mae2.append(mean_absolute_error(val_y1,prediction2[i]))
    mae3.append(mean_absolute_error(val_y1,prediction3[i]))
print(min(mae1),mae1.index(min(mae1)) + 5)
print(min(mae2),mae2.index(min(mae2)) + 5)
print(min(mae3),mae3.index(min(mae3)) + 5)
print(mae1)
print(mae2)
print(mae3)
```

```python
modelA = KNeighborsRegressor(18)
modelB = KNeighborsRegressor(14)
modelC = KNeighborsRegressor(5)
modelA.fit(train_X,train_y1)
modelB.fit(train_X,train_y2)
modelC.fit(train_X,train_y3)
```

```python
test_X = test_data[features]
predictionA = modelA.predict(test_X)
predictionB = modelB.predict(test_X)
predictionC = modelC.predict(test_X)
print(predictionA)
print(predictionB)
print(predictionC)
```
