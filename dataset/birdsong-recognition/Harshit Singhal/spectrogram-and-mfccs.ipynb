```python
import librosa, librosa.display
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
```

```python
pd.set_option("display.max_columns", 40)
```

```python
train_df = pd.read_csv("/kaggle/input/birdsong-recognition/train.csv")
train_df.head()
```

```python
base_dir = "/kaggle/input/birdsong-recognition/train_audio/"
train_df['full_path'] = base_dir + train_df['ebird_code'] + '/' + train_df['filename']
```

# ***Aldfly Bird***

```python
# Lets see about the 'aldfly' bird
file = train_df[train_df['ebird_code'] == 'aldfly'].reset_index()['full_path'][0]
signal, sr = librosa.load(file, sr=40200)
# WAVEPLOT
librosa.display.waveplot(signal, sr = sr)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show()


n_fft = 2048
hop_length = 512 # How much we have to shift
stft = librosa.core.stft(signal, hop_length = hop_length, n_fft = n_fft)
spectrogram = np.abs(stft)

log_spectrogram  = librosa.amplitude_to_db(spectrogram)

# SPECTROFRAM
librosa.display.specshow(spectrogram, sr = sr, hop_length = hop_length)
plt.xlabel("Time")
plt.ylabel("Frequency")
plt.colorbar()
plt.set_cmap("viridis")
plt.show()

# LOG SPECTROGRAM
librosa.display.specshow(log_spectrogram, sr = sr, hop_length = hop_length)
plt.xlabel("Time")
plt.ylabel("Frequency")
plt.set_cmap("viridis")
plt.colorbar()
plt.show()

#MFCCs
MFCCs = librosa.feature.mfcc(signal, n_fft = n_fft, hop_length = hop_length, n_mfcc = 13)
librosa.display.specshow(MFCCs, sr = sr, hop_length = hop_length)
plt.xlabel("Time")
plt.ylabel("MFCCs")
plt.set_cmap("viridis")
plt.colorbar()
plt.show()
```

```python
ipd.Audio(file)
```
