```python
train_data = pd.read_csv("../input/birdsong-recognition/train.csv")
train_data = train_data.drop(["author", "recordist", "license", "rating", "date", "url", "file_type", "background", "primary_label", "description" ], axis=1)

train_data = train_data[["ebird_code", "filename"]]

train_data.head()
```

```python
le = sklearn.preprocessing.LabelEncoder()
train_data["ebird_code"] = le.fit_transform(train_data["ebird_code"])
train_data.head()
```

```python
import os

filepaths = []

directories = sorted(os.listdir('/kaggle/input/birdsong-recognition/train_audio'))

for directory in directories:
    filepath = '/kaggle/input/birdsong-recognition/train_audio/' + directory + "/"
    for dirpath, dirnames, filenames in os.walk(filepath):
        filenames = sorted(filenames)
        for filename in filenames:
            filepaths.append(os.path.join(dirpath, filename))
    
    
train_data['filepath'] = pd.Series(filepaths)

print("complete")
```

```python
pd.set_option('display.max_colwidth', None)
train_data.head()
```

```python
plt.figure(figsize = (10,8))

audio, sr = librosa.load(train_data["filepath"][0], sr=44100)
X = librosa.stft(audio)
Xdb = librosa.amplitude_to_db(abs(X))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='log')
plt.colorbar()
```

```python
import warnings
warnings.filterwarnings("ignore")

cmap = plt.get_cmap('inferno')
plt.figure(figsize = (8,8))

def mp3_to_img(filepath, num):
    audio, sr = librosa.load(filepath, sr=44100)
    plt.specgram(audio, NFFT=2048, Fs=2, Fc=0, noverlap=128, cmap=cmap, sides='default', mode='default', scale='dB');
    plt.axis('off');
    plt.savefig(f'img{num}.png')
    plt.clf()


i = 0
for file in train_data['filepath']:
    mp3_to_img(file, i)
    i+=1


print("complete")
```
