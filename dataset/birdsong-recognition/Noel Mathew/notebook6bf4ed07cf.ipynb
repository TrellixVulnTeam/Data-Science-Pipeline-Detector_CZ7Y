```python
class Site3Dataset:
    def __init__(self, df, path):
        self.df = df
        self.path = path
        self.transforms = [torchaudio.transforms.MelSpectrogram(sample_rate=32000, n_fft=512),
                           torchaudio.transforms.AmplitudeToDB()]
        self.clips = None
    def __len__(self):
        return len(self.df)*2
    
    def __getitem__(self, idx):
        row = self.df.iloc[idx//2]
        if self.clips is not None:
            clips = self.clips
            self.clips = None
            return {
            'audio_id': row['audio_id'],
            'spec': clips,
            }
        row = self.df.iloc[idx//2]
        filename= row['audio_id']+'.mp3'
        site = row['site']
        sr,x  = read(self.path/'test_audio'/filename)
        x = torch.from_numpy(x)
        start_idxs = np.arange(0,len(x))[::sr*5]
        end_idxs = np.arange(0,len(x))[sr*5::sr*5]
        sub_count = len(end_idxs)
        clips = []
        for lo,hi in zip(start_idxs[:sub_count], end_idxs[:sub_count]):
            clip = x[lo:hi]
            clips.append(clip)
        clips = torch.stack(clips).float()
        for tfm in self.transforms:
            clips = tfm(clips)
        self.clips = clips[len(clips)//2:,...]
        
        return {
            'audio_id': row['audio_id'],
            'spec': clips,
        }
```

```python
class Dataset:
    def __init__(self, df, path):
        self.df = df
        self.path = path
        self.transforms = [torchaudio.transforms.MelSpectrogram(sample_rate=32000, n_fft=512),
                           torchaudio.transforms.AmplitudeToDB()]

    def __len__(self):
        return len(self.df)
    
    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        seconds = np.int(row['seconds'].item())
        filename = row['audio_id']+'.mp3'
        site = row['site']
        sr,x  = read(self.path/'test_audio'/filename)
        x = torch.from_numpy(x).float()
        start_idxs = sr*(seconds-5)
        end_idxs = sr*(seconds)
        clip = x[start_idxs:end_idxs]
        for tfm in self.transforms:
            clip = tfm(clip)
                        
        return {
            'audio_id': row['audio_id'],
            'spec': clip,
        }
```
