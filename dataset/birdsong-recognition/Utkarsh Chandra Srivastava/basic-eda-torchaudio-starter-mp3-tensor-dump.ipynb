```python
import pandas as pd 
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
import torch
import torchaudio
import os
import IPython.display as ipd
import numpy as np
from plotly.offline import init_notebook_mode, iplot, plot
import plotly as py
init_notebook_mode(connected=True)
pd.set_option('display.max_columns', None)
```

```python
audio_files=[]
for root,_,files in os.walk("../input/birdsong-recognition/train_audio"):
    for f in files:
        audio_files.append(os.path.join(root,f))
```

```python
df=pd.read_csv("../input/birdsong-recognition/train.csv")
```

# Birds Distribution Over The Globe


```python
df_bird_geo=pd.DataFrame()
df_bird_geo['Bird_Name']=df['species']
df_bird_geo['lattitude']=df['latitude']
df_bird_geo['longitude']=df['longitude']
df_bird_geo['country']=df['country']
df_bird_geo_1=df_bird_geo.drop_duplicates()
```

```python
fig = px.scatter_geo(df_bird_geo_1, lat="lattitude",lon="longitude",
                     color="Bird_Name", # which column to use to set the color of markers
                     hover_name="country", # column added to hover information
                     projection="natural earth")
iplot(fig)
```

# Function To Conver Audio .mp3 / wav  -> Tensor 

```python
def convert_to_tensor(audio_file):
    tensor, _= torchaudio.load(audio_file)
    torch.save(tensor,'_tensor')
    return tensor
```

```python
convert_to_tensor(audio_files[0])
```
