## Test Spectrogram Function

```python
row_id = []

def create_spectrogram(df, start_time, duration, mode):

    if mode== 'test':
        filepath = (test_dir +df[3] +'.mp3') 
    
    try:
        fig = plt.figure(figsize=[2.7,2.7])
        clip, sample_rate = librosa.load(filepath, sr=None, offset=start_time, duration=duration)
        S = librosa.feature.melspectrogram(y=clip, sr=sample_rate)
        librosa.display.specshow(librosa.power_to_db(S, ref=np.max))

        if mode== 'test':
            if not os.path.exists(test_filesave_dir):
                os.makedirs(test_filesave_dir)  
            plt.savefig((test_filesave_dir +df[1] +'.jpg'), bbox_inches='tight',pad_inches=0, facecolor='black')
            
            row_id.append(df[1])
            
        fig.clear()          
        plt.close(fig)
        plt.close()
        plt.close('all')   
        plt.cla()
        fig.clf()
        plt.clf()
        plt.close()
    
    except:
        print("found a broken Audio File")     
```

```python
# Getting Rid of Un-Wanted Warnings when Loading Audio Files
import warnings

warnings.filterwarnings("ignore")
```

```python
for row in test_df.values:
    
    if row[0]=='site_3' :
        start_time= 0
        create_spectrogram(row, start_time, duration= None, mode= 'test')
    else :
        start_time = row[2] - 5
        create_spectrogram(row, start_time, duration= 5, mode= 'test')
```

```python
def predict(path):
    img = cv2.imread(str(path))
    img = cv2.resize(img, (150,150))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = img.astype(np.float32)/255.
    img = np.reshape(img,(1,150,150,3))
    return model.predict(img)
```

```python
try:
    target = []

    for img in row_id:
        target.append(predict(test_filesave_dir +img +'.jpg'))
    
except:
    pass
```

## Submission

```python
try:   
    pred_class = []
    for row in target:
        if row.max()< 0.45:
            pred_class.append('nocall')
        else:
            pred_class.append(rev_bird_dict[np.argmax(row, axis=1)[0]])
    
    predict_df = pd.DataFrame(row_id)
    predict_df.columns = ['row_id']
    predict_df['birds'] = pred_class
    
except:
    predict_df = pd.read_csv('../input/birdsong-recognition/sample_submission.csv')
```
