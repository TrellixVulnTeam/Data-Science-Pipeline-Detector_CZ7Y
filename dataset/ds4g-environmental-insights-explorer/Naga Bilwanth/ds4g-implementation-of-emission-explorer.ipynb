<p id="backgroundStudy" style="color:#703bdb; font-size:20px">1. Background</p>
<hr size="30"> 

&emsp;&emsp; By raising interest in healthy living worldwide, there has been an aggregate demand for more precise measurements of the concentrations of air pollutants such as NO2, CO2. For this competition, focus is mainly on NO2 emissions. “Nitrogen oxides” (NOx) is the term used to describe the sum of nitric oxide (NO), nitrogen dioxide(NO2), and other oxides of nitrogen. Most airborne NOx comes from combustion-related emissions sources of human origin, primarily fossil fuel combustion in electric utilities, high-temperature operations at other industrial sources, and operation of motor vehicles. In certain areas, studying the characterstics and sources of air pollutants by region could improve the efficiency of environmental guidelines applied in accordance with the environmental characteristics of specific regions. 

<p id="problemStatement" style="color:#703bdb; font-size:20px">2. Problem Statement</p>
<hr size="30"> 

&emsp;&emsp;Current emission factors are based on time-consuming data collection and may include errors derived from lack of access to granular datasets, inability to refresh data on a frequent basis, overly general modeling assumptions, and inaccurate reporting of emissions sources like fuel consumption. This begs the question: What if there was a different way to calculate or measure emissions factors? We’re challenging the Kaggle community to see if it’s possible to use remote sensing techniques for better model emissions factors. You will develop a methodology to calculate an average historical emissions factor for electricity generation in a sub-national region.

For analysis purpose, received data for Puerto Rico region from different sources

* <a href="https://developers.google.com/earth-engine/datasets/catalog/WRI_GPPD_power_plants" >Global Power Plants data</a> : 
    This provide estimated generation capacity of power plant to calculate emission factor in Puerto Rico region

* <a href="https://developers.google.com/earth-engine/datasets/catalog/COPERNICUS_S5P_OFFL_L3_NO2"> 
    Sentinel 5P OFFL NO2 satelite images <a/> <font color="black">  by EU/ESA/Copernicus has been used to observe air pollutants. Sentinel-5 Precursor is the first mission of the Copernicus Programme dedicated to monitoring air pollution. Its instrument is an ultraviolet, visible, near and short-wavelength infrared spectrometer called Tropomi. The Satellite is built on a hexagonal Astrobus L 250 satellite bus equipped with S- and X-band communication antennas, three foldable solar panels generating 1500 watts and hydrazine thrusters for station-keeping.

* <a href='https://developers.google.com/earth-engine/datasets/catalog/NOAA_GFS0P25'>Global Forecast system 384-Hour Predicted Atmosphere Data </a> by NOAA/NCEP/EMC. This data is used to observe behaviour of No2 emissions based on weather changes.

<p id="idealSolution" style="color:#703bdb; font-size:20px">5. An Ideal Solution : Key Characteristics</p>

Apart from these challenges, the problem statement of this project requires to design a solution with

**Hassle-Free Calculations:** A lot of mathematical steps are to be performed in this project in order calculate emission factor and comparing behaviour of emissions with weather factors. So the solution should not require too many changes.

**Need for Scalability**: This project needs to be scaled at a very high level, so one needs to design a solution that is scalable in nature.

**Least Human Intervenion:** While working with this type of data, a lot of human interventation can be involved but that solution will not be the best one. So the solution should be designed with least manual intervenion.

**Quickly Give Insights:** A best solution should be able to quickly give insights and report emission factor

<p id="solutionOverview" style="color:#703bdb; font-size:20px">6. The Solution - Overview</p>
<hr size="30">

&emsp;&emsp;Keeping all these challenges and charateristics in mind, The best solution from a data science perspective is to first bring a structure in the data before performing any statistical modelling or analysis. Later, cleaned data can be analysed using an Analytics Framework in order to provide actionable insights and well defined relevant reports. The solution is composed of **an end to end automation data processing and analysis pipeline** which takes care of both the points : Data Processing and Data Analysis. The complete pipeline has three major components or parts named as Component A, Component B and Component C:

* Component A : Data Integration and Standardization
* Component B : Emissions / Weather Data Processing
* Component C : Analysis Framework
