```python
players_df = pd.read_csv('/kaggle/input/nfl-big-data-bowl-2022/players.csv')
players_df.head()
```

```python
players_df = players_df[players_df.Position != "WR"]
players_df = players_df[players_df.Position != "CB"]
players_df = players_df[players_df.Position != "DE"]
players_df = players_df[players_df.Position != "OLB"]
players_df = players_df[players_df.Position != "TE"]
players_df = players_df[players_df.Position != "T"]
players_df = players_df[players_df.Position != "RB"]
players_df = players_df[players_df.Position != "DT"]
players_df = players_df[players_df.Position != "ILB"]
players_df = players_df[players_df.Position != "FS"]
players_df = players_df[players_df.Position != "SS"]
players_df = players_df[players_df.Position != "C"]
players_df = players_df[players_df.Position != "NT"]
players_df = players_df[players_df.Position != "DB"]
players_df = players_df[players_df.Position != "LB"]
players_df = players_df[players_df.Position != "MLB"]
players_df = players_df[players_df.Position != "FB"]
players_df = players_df[players_df.Position != "OT"]
players_df = players_df[players_df.Position != "QB"]
players_df = players_df[players_df.Position != "S"]
players_df = players_df[players_df.Position != "OG"]
players_df = players_df[players_df.Position != "HB"]

# # K H P LS G KR/PR
# players_df['Position'].value_counts()
```

```python
height_dict = {
    "5-10": 178,
    "5-11": 180,
    "5-6": 168,
    "5-7": 170,
    "5-8": 173,
    "5-9": 175,
    "6-0": 183,
    "6-1": 185,
    "6-2": 188,
    "6-3": 190,
    "6-4": 193,
    "6-5": 196,
    "6-6": 198,
    "6-7": 201,
    "6-8": 203,
    "6-9": 206,
    "66": 168,
    "67": 170,
    "68": 173,
    "69": 175,
    "70": 178,
    "71": 180,
    "72": 183,
    "73": 185,
    "74": 188,
    "75": 191,
    "76": 193,
    "77": 196,
    "78": 198,
    "79": 201
}
```

```python
players_df["height_cm"] = players_df["height"].replace(height_dict)
players_df["height_m2"] = players_df["height_cm"]*0.01*players_df["height_cm"]*0.01
players_df["weight_kg"] = players_df["weight"]*0.45359237
players_df
```

```python
import plotly.express as px
fig = px.box(players_df,x="Position",y="height_cm",points = "all")
fig.show()
```

```python
fig = px.box(players_df,x="Position",y="weight_kg", points = "all")
fig.show()
```

```python
players_df["bmi"] = players_df["weight_kg"]/players_df["height_m2"]
fig = px.box(players_df,x="Position",y="bmi", points = "all")
fig.show()
```

```python
players_df["nomalized_bmi"]=(players_df["bmi"]-players_df["bmi"].min())/(players_df["bmi"].max()-players_df["bmi"].min())*100
players_df
```

```python


import random
color_dict = {
    "G": 'red',
    "K": 'limegreen',
    "P": 'royalblue',
    "LS": 'purple',
    "WR": "#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)]),
    "CB": "#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)]),
    "DE": "#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)]),
    "OLB": "#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)]),
    "TE": "#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)]),
    "T": "#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)]),
    "RB": "#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)]),
    "DT": "#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)]),
    "ILB": "#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)]),
    "FS": "#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)]),
    "SS": "#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)]),
    "C": "#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)]),
    "NT": "#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)]),
    "DB": "#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)]),
    "MLB": "#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)]),
    "FB": "#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)]),
    "OT": "#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)]),
    "QB": "#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)]),
    "S": "#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)]),
    "OG": "#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)]),
    "HB": "#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)]),
    "LB": "#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)])
}
players_df["color"] = players_df["Position"].replace(color_dict)
players_df
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10, 6))
plt.scatter(players_df["height_cm"], players_df["weight_kg"],c=players_df["color"], s=players_df["nomalized_bmi"],alpha=0.5)
plt.xlabel('height_cm')
plt.ylabel('weight_kg')


import matplotlib.patches as mpatches
from collections import OrderedDict
# plot the color description bar
colorlist = zip(players_df['Position'], players_df['color'])
colorlist = list(OrderedDict.fromkeys(colorlist))
handles = [mpatches.Patch(color=colour, label=label) for label, colour in colorlist]
colorlist = [i[0] for i in colorlist]
plt.legend(handles, colorlist, bbox_to_anchor=(1.1, 1))

plt.show()
```
