# #  **Players Data**

```python
players = pd.read_csv("../input/nfl-big-data-bowl-2022/players.csv")
```

```python
players.head(n=5)
```

```python
# players data description
players.describe()
players.info()
players.shape 
players.columns
players.dtypes
players.isnull().sum()
```

```python
players["birthYear"] = 0
players["birthMonth"] = 0
#There are NA values in birthDate so that we should drop them
players.dropna(subset=["birthDate"], inplace=True)
for idx, row in players.iterrows():
    if len(row['birthDate'].split('/')) == 3: # 05/17/1994 
        players.loc[idx, 'birthYear'] = row['birthDate'].split('/')[2]
        players.loc[idx, 'birthMonth'] = row['birthDate'].split('/')[0]
        
    elif len(row['birthDate'].split('-')) == 3: # 1995-05-05
        players.loc[idx, 'birthYear'] = row['birthDate'].split('-')[0]
        players.loc[idx, 'birthMonth'] = row['birthDate'].split('-')[1]
```

```python
players.isnull().sum()
```

```python
players.tail(n=5)
```

```python
players.head()
```

```python
# lets create a new column for age
players['age'] = 0
for idx, row in players.iterrows():
    players.loc[idx, 'age'] = 2020 - int(row['birthYear'])
```

```python
players.head()
```

```python
# convert heights to cm and weights to kg
players_heights = players["height"] # Get the Height data from DataFrame
players_heights = players_heights.apply(lambda x: x.split("-")) # Split the heights by hyphen ("-")

# Convert Heights to Centimeters and add them to DataFrame
players["height"] = players_heights.apply(lambda x: int(x[0]) * 12 + int(x[1]) if len(x) == 2 else int(x[0])) * 2.54

# Convert Weights to Kilograms and them to DataFrame
players["weight"] = round(players.weight * 0.453592, 2)

```

# # ****Games Data****

```python
# games data 
games = pd.read_csv('../input/nfl-big-data-bowl-2022/games.csv')
```

```python
games.head(n=5)
```

```python
#games data description
games.describe()
games.info()
games.shape
games.columns
games.dtypes
games.isnull().sum()
games.isnull().sum().sum()
```

```python
# add column month of year and day of week
games['month'] = pd.to_datetime(games['gameDate']).dt.month_name()
games['day_of_week'] = pd.to_datetime(games['gameDate']).dt.day_name()
games['day'] = games['gameDate'].apply(lambda x: int(x.split('/')[1]))
games.head()
```

```python
#writing function for perentage  at the top of bar 
def write_percentage(ax,total_size):
    for p in ax.patches:
        x=p.get_x()+p.get_width()/2
        y=p.get_y()+p.get_height()
        val=int(p.get_height()/total_size*100)
        ax.annotate('{}%'.format(val), (x,y), ha='center', va='bottom', fontsize=10)
```

```python
 #games data visulaization count per season with perentage at top of bar 
sns.set_style('whitegrid')
sns.set_context('notebook', font_scale=1.5, rc={"lines.linewidth": 2.5})
write_percentage(sns.countplot(x='season', data=games , palette='Set2', order=games.season.unique()),games.season.count())
```

```python
# games data visulaization Number of games per month with percentage at top of bar 
mpl.rc('font', size=15)
plt.figure(figsize=(8, 6))
ax=(sns.countplot(x='month', data=games , palette='Set2', order=games.month.unique()))
write_percentage(ax,games.month.count())
ax.set_title('Number of games per month')
plt.show()
```

```python
#games data visulaization Number of games per day with percentage at top of bar with title number of games per day
mpl.rc('font', size=15)
plt.figure(figsize=(8, 6))
ax= sns.countplot(x='day_of_week', data=games , palette='Set2', order=games.day_of_week.unique())
write_percentage(ax,games.day_of_week.count())
ax.set_title('Number of games per day')
plt.show();
```

```python
# games data visulaization count per week
sns.set_style('whitegrid')
sns.set_context('notebook', font_scale=1.5, rc={"lines.linewidth": 2.5})
sns.countplot(x='week', data=games , palette='Set2', order=games.week.unique())
plt.show();
```

```python
#Number of games per day with percentage at top of bar with title number of games per day
mpl.rc('figure', figsize=(10, 6))
plt.figure(figsize=(15, 7))
ax =sns.countplot(x='day', data=games , palette='Set2', order=games.day.unique())
write_percentage(ax,games.day.count())
ax.set_title('Number of games per day')
plt.show();
```

```python
#Number of games for gameTimeEastern and gameTimeLocal in ascending order with percentage at top of bar with title number of games per gameTimeEastern
mpl.rc('figure', figsize=(10, 6))
plt.figure(figsize=(15, 7))
ax=sns.countplot(x='gameTimeEastern', data=games , palette='Set2', order=games.gameTimeEastern.unique())
write_percentage(ax,games.gameTimeEastern.count())
ax.set_title('Number of games per gameTimeEastern')
ax.tick_params('x', labelrotation=30)
plt.show();
```

```python
#Plotting Number of Games played in each month for each season
sns.set_style('whitegrid')
sns.set_context('notebook', font_scale=1.5, rc={"lines.linewidth": 2.5})

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 2, box.height * 2])

ax.set_title('Number of Games played in each month for each season')
ax.set_xlabel('Season')
ax.set_ylabel('Number of Games')

ax=sns.countplot(x='season', hue='month', data=games, palette='Set2', order=games.season.unique())

plt.show();
```
