```python
height_df = players_df['height'].str.split('-',expand=True)

height_df.columns = [
    'feet', 
    'inches'
]

height_df.loc[(height_df['inches'].notnull()), 'feet'] = height_df[height_df['inches'].notnull()]['feet'].astype(np.int16) * 12 + height_df[height_df['inches'].notnull()]['inches'].astype(np.int16)
```

```python
players_df['height'] = height_df['feet']
players_df['height'] = players_df['height'].astype(np.float32)
players_df['height'] /= 12
```

### Players Date of Birth Analysis 

```python
players_df["birthYear"] = 0
players_df["birthMonth"] = 0
#There are NA values in birthDate so that we should drop them
players_df.dropna(subset=["birthDate"], inplace=True)
for idx, row in players_df.iterrows():
    if len(row['birthDate'].split('/')) == 3: 
        players_df.loc[idx, 'birthYear'] = row['birthDate'].split('/')[2]
        players_df.loc[idx, 'birthMonth'] = row['birthDate'].split('/')[0]
        
    elif len(row['birthDate'].split('-')) == 3: 
        players_df.loc[idx, 'birthYear'] = row['birthDate'].split('-')[0]
        players_df.loc[idx, 'birthMonth'] = row['birthDate'].split('-')[1]
```

```python
plt.figure(figsize=(16, 10))
ax=sns.countplot(x="birthYear",data=players_df,order=players_df.birthYear.value_counts().index,palette="Dark2_r",linewidth=3)
ax.bar_label(ax.containers[0])
plt.title("Birth Year of the players",font="Serif", size=20)
plt.show()
```

```python
data = go.Pie(
values= players_df["birthMonth"].value_counts(),
labels= players_df["birthMonth"].value_counts().index,
)
layout = go.Layout(
title=dict(text = "Birth Month for players",x=0.46,y=0.95,font_size=20)
)
fig = go.Figure(data=data,layout=layout)
fig.show()
```
