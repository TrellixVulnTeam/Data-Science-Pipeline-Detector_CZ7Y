# Project Output

**Every team who won the Super Bowl** **MADE ALL** *of their field goal attempts in the 4th quarter of a regular season game (2018-2020) on 1st, 2nd, and 4th down when the pre-snap home score minus visitor score was within 2 points.*

**Every team who won the Super Bowl** **MISSED ALL** *of their field goal attempts in the 4th quarter of a regular season game (2018-2020) on 3rd down when the pre-snap home score minus visitor score was within 2 points.*

Average kick length of these attempts was 38.64 yards and 11/14 fields goals were made or approximately 79% *(part of the 4th Quarter Kicking Index formula shared later in this report).*

# Conclusion

The goal of this report is to emphasize the impact that you can find by starting to sort a large dataset and telling a clear story. There are certainly more complex models that can be developed from this data, however, the foundation of problem-solving that can be measured week-to-week and game-by-game via NFL field goal kicking is still found here. 

Overall, this is what the concept of the *4th Quarter Kicking Index* could look like no matter what iterative software you prefer to develop the final output in with as a possible formula:

- Select all NFL regular season game data within last three years of your current NFL season and filter by field goals attempted only in the 4th quarter, then ensure pre-snap game score is in no more than a 2-point margin (you can break this down by team and/or player for deeper comparison)

**4th Quarter Kicking Index = (4th quarter field goals made divided by attempted) / .8 (rounded up from .79 in the Project Output section above)**

**
**Team 4th Quarter Kicking Index scores will range between 0 and 1.25 or 125%**

** 
**A team with a 4th Quarter Kicking Index score of greater than or equal to 1 or 100% may have a higher chance of winning the Super Bowl within the next three seasons than those teams below this score**

**i.e. Team A makes 4/8 field goals in the fourth quarter for the regular season when the score is a 2-point margin or less and Team B makes 7/8 field goals. Team B's 4th Quarter Kicking Index is greater at ((7/8)/.8)) = 1.09 or 109% than Team A's 4th Quarter Kicking Index at ((4/8)/.8)) = .625 or 62.5%, meaning Team B may *possibly* have a 46% higher chance of winning the Super Bowl in the next three years than Team A if you *only* base this on 4th Quarter Kicking Index scores alone.**

*It is honestly crazy to think right now that a team will always win a Super Bowl in the next three years that makes nearly .8 (from the 4th Quarter Kicking Index formula), 8/10, or 80% of their field goals in the 4th quarter during a regular season when the game is on the line with a pre-snap score difference of 2 or less points. We will see in future seasons how this data looks and how making 3 points tells another story.*
