**<p id='1'><li>Function to Load Video File</li></p>**

```python
def playVideo(videoFile):
    html = ''
    video = open(videoFile,'rb').read()
    src = 'data:video/mp4;base64,' + b64encode(video).decode()
    html += '<video width=1100 controls autoplay loop><source src="%s" type="video/mp4"></video>' % src 
    return HTML(html)
```

# Data Loading

```python
extraPointsKC2018 = pd.read_csv('../input/kansas-city-chiefs-special-plays-2018-to-2020/KC_ExtraPoint_2018.csv')
extraPointsKC2019 = pd.read_csv('../input/kansas-city-chiefs-special-plays-2018-to-2020/KC_ExtraPoint_2019.csv')
extraPointsKC2020 = pd.read_csv('../input/kansas-city-chiefs-special-plays-2018-to-2020/KC_ExtraPoint_2020.csv')

fieldGoalsKC2018 = pd.read_csv('../input/kansas-city-chiefs-special-plays-2018-to-2020/KC_FiledGoals_2018.csv')
fieldGoalsKC2019 = pd.read_csv('../input/kansas-city-chiefs-special-plays-2018-to-2020/KC_FiledGoals_2019.csv')
fieldGoalsKC2020 = pd.read_csv('../input/kansas-city-chiefs-special-plays-2018-to-2020/KC_FiledGoals_2020.csv')

kickoffsKC2018 = pd.read_csv('../input/kansas-city-chiefs-special-plays-2018-to-2020/KC_Kickoffs_2018.csv')
kickoffsKC2019 = pd.read_csv('../input/kansas-city-chiefs-special-plays-2018-to-2020/KC_Kickoffs_2019.csv')
kickoffsKC2020 = pd.read_csv('../input/kansas-city-chiefs-special-plays-2018-to-2020/KC_Kickoffs_2020.csv')

puntsKC2018 = pd.read_csv('../input/kansas-city-chiefs-special-plays-2018-to-2020/KC_Punts_2018.csv')
puntsKC2019 = pd.read_csv('../input/kansas-city-chiefs-special-plays-2018-to-2020/KC_Punts_2019.csv')
puntsKC2020 = pd.read_csv('../input/kansas-city-chiefs-special-plays-2018-to-2020/KC_Punts_2020.csv')

pFFData = pd.read_csv('../input/nfl-big-data-bowl-2022/PFFScoutingData.csv')
pd.options.display.max_rows = 99999999
```

<a id="3"></a> <br>
# Data Preparation for Kick of Return

```python
KCKickoffReturn2018Above25Yards = kickoffsKC2018[(kickoffsKC2018.Result == 'Kickoff - Return') & (kickoffsKC2018.kickReturnYardage > 25)]
KCKickoffReturn2019Above25Yards = kickoffsKC2019[(kickoffsKC2019.Result == 'Kickoff - Return') & (kickoffsKC2019.kickReturnYardage > 25)]
KCKickoffReturn2020Above25Yards = kickoffsKC2020[(kickoffsKC2020.Result == 'Kickoff - Return') & (kickoffsKC2020.kickReturnYardage > 25) & (kickoffsKC2020.playResult > 0)]

KCKickoffReturn2018Below25Yards = kickoffsKC2018[(kickoffsKC2018.Result == 'Kickoff - Return') & (kickoffsKC2018.kickReturnYardage <= 25)]
KCKickoffReturn2019Below25Yards = kickoffsKC2019[(kickoffsKC2019.Result == 'Kickoff - Return') & (kickoffsKC2019.kickReturnYardage <= 25)]
KCKickoffReturn2020Below25Yards = kickoffsKC2020[(kickoffsKC2020.Result == 'Kickoff - Return') & (kickoffsKC2020.kickReturnYardage <= 25)]

KCKickoffReturn2020LossYards = kickoffsKC2020[(kickoffsKC2020.Result == 'Kickoff - Return') & (kickoffsKC2020.playResult < 0)]
```

<a id="4"></a> <br>
# Merging Kickoff Return Data with Pro Football Focus Data

```python
# Merged Data with Kickoff Return Above 25 Yards.
mergedKCKickoffReturn2018Above25Yards = pd.merge(KCKickoffReturn2018Above25Yards, 
                                                 pFFData,  
                                                 how='left', 
                                                 left_on=['gameId','playId'], 
                                                 right_on = ['gameId','playId']
                                                )

mergedKCKickoffReturn2019Above25Yards = pd.merge(KCKickoffReturn2019Above25Yards, 
                                                 pFFData,  
                                                 how='left', 
                                                 left_on=['gameId','playId'], 
                                                 right_on = ['gameId','playId']
                                                )

mergedKCKickoffReturn2020Above25Yards = pd.merge(KCKickoffReturn2020Above25Yards, 
                                                 pFFData,  
                                                 how='left', 
                                                 left_on=['gameId','playId'], 
                                                 right_on = ['gameId','playId']
                                                )

# Merged Data with Kickoff Return Below 25 Yards.
mergedKCKickoffReturn2018Below25Yards = pd.merge(KCKickoffReturn2018Below25Yards, 
                                                 pFFData,  
                                                 how='left', 
                                                 left_on=['gameId','playId'], 
                                                 right_on = ['gameId','playId']
                                                )

mergedKCKickoffReturn2019Below25Yards = pd.merge(KCKickoffReturn2019Below25Yards, 
                                                 pFFData,  
                                                 how='left', 
                                                 left_on=['gameId','playId'], 
                                                 right_on = ['gameId','playId']
                                                )

mergedKCKickoffReturn2020Below25Yards = pd.merge(KCKickoffReturn2020Below25Yards, 
                                                 pFFData,  
                                                 how='left', 
                                                 left_on=['gameId','playId'], 
                                                 right_on = ['gameId','playId']
                                                )

# Merged Data with Kickoff Return Lossing Yards.
mergedKCKickoffReturn2020LossYards = pd.merge(KCKickoffReturn2020LossYards, 
                                                 pFFData,  
                                                 how='left', 
                                                 left_on=['gameId','playId'], 
                                                 right_on = ['gameId','playId']
                                                )
```

<a id="5"></a> <br>
# Grouping DataFrames

```python
# Above 25 Yards 2018
KCKickoffReturn2018Above25YardsSeries = KCKickoffReturn2018Above25Yards.groupby(
    ['season','possessionTeam','gameId','playId','quarter','down','yardsToGo','yardlineSide','Result',
     'gameClock','yardlineNumber','kickReturnYardage','playResult','preSnapHomeScore','preSnapVisitorScore',
     'event','homeTeamAbbr','visitorTeamAbbr','week'],as_index=True).size()
KCKickoffReturn2018Above25YardsFrame = pd.DataFrame(KCKickoffReturn2018Above25YardsSeries)

mergedKCKickoffReturn2018Above25YardsSeries = mergedKCKickoffReturn2018Above25Yards.groupby(
    ['season','kickoffReturnFormation','tackler','gameId','playId','returnDirectionIntended',
     'returnDirectionActual','kickReturnYardage','homeTeamAbbr','visitorTeamAbbr','week'],as_index=True).size()
mergedKCKickoffReturn2018Above25YardsFrame = pd.DataFrame(mergedKCKickoffReturn2018Above25YardsSeries)

# Above 25 Yards 2019
KCKickoffReturn2019Above25YardsSeries = KCKickoffReturn2019Above25Yards.groupby(['season','possessionTeam','gameId','playId','quarter','down','yardsToGo','yardlineSide','Result','gameClock',
                                   'yardlineNumber','kickReturnYardage','playResult','preSnapHomeScore','preSnapVisitorScore','event','homeTeamAbbr','visitorTeamAbbr',
                                   'week'],as_index=True).size()
KCKickoffReturn2019Above25YardsFrame = pd.DataFrame(KCKickoffReturn2019Above25YardsSeries)

mergedKCKickoffReturn2019Above25YardsSeries = mergedKCKickoffReturn2019Above25Yards.groupby(
    ['season','kickoffReturnFormation','tackler','gameId','playId','returnDirectionIntended',
     'returnDirectionActual','kickReturnYardage','homeTeamAbbr','visitorTeamAbbr','week'],as_index=True).size()
mergedKCKickoffReturn2019Above25YardsFrame = pd.DataFrame(mergedKCKickoffReturn2019Above25YardsSeries)

# Above 25 Yards 2020
KCKickoffReturn2020Above25YardsSeries = KCKickoffReturn2020Above25Yards.groupby(['season','possessionTeam','gameId','playId','quarter','down','yardsToGo','yardlineSide','Result','gameClock',
                                   'yardlineNumber','kickReturnYardage','playResult','preSnapHomeScore','preSnapVisitorScore','event','homeTeamAbbr','visitorTeamAbbr',
                                   'week'],as_index=True).size()
KCKickoffReturn2020Above25YardsFrame = pd.DataFrame(KCKickoffReturn2020Above25YardsSeries)

mergedKCKickoffReturn2020Above25YardsSeries = mergedKCKickoffReturn2020Above25Yards.groupby(
    ['season','kickoffReturnFormation','tackler','gameId','playId','returnDirectionIntended',
     'returnDirectionActual','kickReturnYardage','homeTeamAbbr','visitorTeamAbbr','week'],as_index=True).size()
mergedKCKickoffReturn2020Above25YardsFrame = pd.DataFrame(mergedKCKickoffReturn2020Above25YardsSeries)

# Below 25 Yards 2018
KCKickoffReturn2018Below25YardsSeries = KCKickoffReturn2018Below25Yards.groupby(['season','possessionTeam','gameId','playId','quarter','down','yardsToGo','yardlineSide','Result','gameClock',
                                   'yardlineNumber','kickReturnYardage','playResult','preSnapHomeScore','preSnapVisitorScore','event','homeTeamAbbr','visitorTeamAbbr',
                                   'week'],as_index=True).size()
KCKickoffReturn2018Below25YardsFrame = pd.DataFrame(KCKickoffReturn2018Below25YardsSeries)

mergedKCKickoffReturn2018Below25YardsSeries = mergedKCKickoffReturn2018Below25Yards.groupby(
    ['season','kickoffReturnFormation','tackler','gameId','playId','returnDirectionIntended',
     'returnDirectionActual','kickReturnYardage','homeTeamAbbr','visitorTeamAbbr','week'],as_index=True).size()
mergedKCKickoffReturn2018Below25YardsFrame = pd.DataFrame(mergedKCKickoffReturn2018Below25YardsSeries)

# Below 25 Yards 2019
KCKickoffReturn2019Below25YardsSeries = KCKickoffReturn2019Below25Yards.groupby(['season','possessionTeam','gameId','playId','quarter','down','yardsToGo','yardlineSide','Result','gameClock',
                                   'yardlineNumber','kickReturnYardage','playResult','preSnapHomeScore','preSnapVisitorScore','event','homeTeamAbbr','visitorTeamAbbr',
                                   'week'],as_index=True).size()
KCKickoffReturn2019Below25YardsFrame = pd.DataFrame(KCKickoffReturn2019Below25YardsSeries)

mergedKCKickoffReturn2019Below25YardsSeries = mergedKCKickoffReturn2019Below25Yards.groupby(
    ['season','kickoffReturnFormation','tackler','gameId','playId','returnDirectionIntended',
     'returnDirectionActual','kickReturnYardage','homeTeamAbbr','visitorTeamAbbr','week'],as_index=True).size()
mergedKCKickoffReturn2019Below25YardsFrame = pd.DataFrame(mergedKCKickoffReturn2019Below25YardsSeries)

# Below 25 Yards 2020
KCKickoffReturn2020Below25YardsSeries = KCKickoffReturn2020Below25Yards.groupby(['season','possessionTeam','gameId','playId','quarter','down','yardsToGo','yardlineSide','Result','gameClock',
                                   'yardlineNumber','kickReturnYardage','playResult','preSnapHomeScore','preSnapVisitorScore','event','homeTeamAbbr','visitorTeamAbbr',
                                   'week'],as_index=True).size()
KCKickoffReturn2020Below25YardsFrame = pd.DataFrame(KCKickoffReturn2020Below25YardsSeries)

mergedKCKickoffReturn2020Below25YardsSeries = mergedKCKickoffReturn2020Below25Yards.groupby(
    ['season','kickoffReturnFormation','tackler','gameId','playId','returnDirectionIntended',
     'returnDirectionActual','kickReturnYardage','homeTeamAbbr','visitorTeamAbbr','week'],as_index=True).size()
mergedKCKickoffReturn2020Below25YardsFrame = pd.DataFrame(mergedKCKickoffReturn2020Below25YardsSeries)

# Loosing Yards 2020
KCKickoffReturn2020LossYardsSeries = KCKickoffReturn2020LossYards.groupby(['season','possessionTeam','gameId','playId','quarter','down','yardsToGo','yardlineSide','Result','gameClock',
                                   'yardlineNumber','kickReturnYardage','playResult','preSnapHomeScore','preSnapVisitorScore','event','homeTeamAbbr','visitorTeamAbbr',
                                   'week'],as_index=True).size()
KCKickoffReturn2020LossYardsFrame = pd.DataFrame(KCKickoffReturn2020LossYardsSeries)

mergedKCKickoffReturn2020LossYardsSeires = mergedKCKickoffReturn2020LossYards.groupby(
    ['season','kickoffReturnFormation','specialTeamsSafeties','gameId','playId','returnDirectionIntended',
     'returnDirectionActual','kickReturnYardage','homeTeamAbbr','visitorTeamAbbr','week'],as_index=True).size()
mergedKCKickoffReturn2020LossYardsFrame = pd.DataFrame(mergedKCKickoffReturn2020LossYardsSeires)
```
