# Data Loading

```python
extraPointsTEN2018 = pd.read_csv('../input/tennessee-titans-special-play-2018-to-2019/TEN_ExtraPoint_2018.csv')
extraPointsTEN2019 = pd.read_csv('../input/tennessee-titans-special-play-2018-to-2019/TEN_ExtraPoint_2019.csv')
extraPointsTEN2020 = pd.read_csv('../input/tennessee-titans-special-play-2018-to-2019/TEN_ExtraPoint_2020.csv')

fieldGoalsTEN2018 = pd.read_csv('../input/tennessee-titans-special-play-2018-to-2019/TEN_FiledGoals_2018.csv')
fieldGoalsTEN2019 = pd.read_csv('../input/tennessee-titans-special-play-2018-to-2019/TEN_FiledGoals_2019.csv')
fieldGoalsTEN2020 = pd.read_csv('../input/tennessee-titans-special-play-2018-to-2019/TEN_FiledGoals_2020.csv')

kickoffsTEN2018 = pd.read_csv('../input/tennessee-titans-special-play-2018-to-2019/TEN_Kickoffs_2018.csv')
kickoffsTEN2019 = pd.read_csv('../input/tennessee-titans-special-play-2018-to-2019/TEN_Kickoffs_2019.csv')
kickoffsTEN2020 = pd.read_csv('../input/tennessee-titans-special-play-2018-to-2019/TEN_Kickoffs_2020.csv')

puntsTEN2018 = pd.read_csv('../input/tennessee-titans-special-play-2018-to-2019/TEN_Punts_2018.csv')
puntsTEN2019 = pd.read_csv('../input/tennessee-titans-special-play-2018-to-2019/TEN_Punts_2019.csv')
puntsTEN2020 = pd.read_csv('../input/tennessee-titans-special-play-2018-to-2019/TEN_Punts_2020.csv')

pFFData = pd.read_csv('../input/nfl-big-data-bowl-2022/PFFScoutingData.csv')
pd.options.display.max_rows = 99999999
```

<a id="3"></a> <br>
# Data Preparation for Kick of Return

```python
TENKickoffReturn2018Above25Yards = kickoffsTEN2018[(kickoffsTEN2018.Result == 'Kickoff - Return') & (kickoffsTEN2018.kickReturnYardage > 25) & (kickoffsTEN2018.playResult > 0)]
TENKickoffReturn2019Above25Yards = kickoffsTEN2019[(kickoffsTEN2019.Result == 'Kickoff - Return') & (kickoffsTEN2019.kickReturnYardage > 25)]
TENKickoffReturn2020Above25Yards = kickoffsTEN2020[(kickoffsTEN2020.Result == 'Kickoff - Return') & (kickoffsTEN2020.kickReturnYardage > 25)]

TENKickoffReturn2018Below25Yards = kickoffsTEN2018[(kickoffsTEN2018.Result == 'Kickoff - Return') & (kickoffsTEN2018.kickReturnYardage <= 25)]
TENKickoffReturn2019Below25Yards = kickoffsTEN2019[(kickoffsTEN2019.Result == 'Kickoff - Return') & (kickoffsTEN2019.kickReturnYardage <= 25)]
TENKickoffReturn2020Below25Yards = kickoffsTEN2020[(kickoffsTEN2020.Result == 'Kickoff - Return') & (kickoffsTEN2020.kickReturnYardage <= 25)]

TENKickoffReturn2018LossYards = kickoffsTEN2018[(kickoffsTEN2018.Result == 'Kickoff - Return') & (kickoffsTEN2018.playResult < 0)]
```

<a id="4"></a> <br>
# Merging Kickoff Return Data with Pro Football Focus Data

```python
# Merged Data with Kickoff Return Above 25 Yards.
mergedTENKickoffReturn2018Above25Yards = pd.merge(TENKickoffReturn2018Above25Yards, 
                                                 pFFData,  
                                                 how='left', 
                                                 left_on=['gameId','playId'], 
                                                 right_on = ['gameId','playId']
                                                )

mergedTENKickoffReturn2019Above25Yards = pd.merge(TENKickoffReturn2019Above25Yards, 
                                                 pFFData,  
                                                 how='left', 
                                                 left_on=['gameId','playId'], 
                                                 right_on = ['gameId','playId']
                                                )

mergedTENKickoffReturn2020Above25Yards = pd.merge(TENKickoffReturn2020Above25Yards, 
                                                 pFFData,  
                                                 how='left', 
                                                 left_on=['gameId','playId'], 
                                                 right_on = ['gameId','playId']
                                                )

# Merged Data with Kickoff Return Below 25 Yards.
mergedTENKickoffReturn2018Below25Yards = pd.merge(TENKickoffReturn2018Below25Yards, 
                                                 pFFData,  
                                                 how='left', 
                                                 left_on=['gameId','playId'], 
                                                 right_on = ['gameId','playId']
                                                )

mergedTENKickoffReturn2019Below25Yards = pd.merge(TENKickoffReturn2019Below25Yards, 
                                                 pFFData,  
                                                 how='left', 
                                                 left_on=['gameId','playId'], 
                                                 right_on = ['gameId','playId']
                                                )

mergedTENKickoffReturn2020Below25Yards = pd.merge(TENKickoffReturn2020Below25Yards, 
                                                 pFFData,  
                                                 how='left', 
                                                 left_on=['gameId','playId'], 
                                                 right_on = ['gameId','playId']
                                                )

# Merged Data with Kickoff Return Lossing Yards.
mergedTENKickoffReturn2018LossYards = pd.merge(TENKickoffReturn2018LossYards, 
                                                 pFFData,  
                                                 how='left', 
                                                 left_on=['gameId','playId'], 
                                                 right_on = ['gameId','playId']
                                                )
```

<a id="5"></a> <br>
# Grouping DataFrames

```python
# Above 25 Yards 2018
TENKickoffReturn2018Above25YardsSeries = TENKickoffReturn2018Above25Yards.groupby(['season','possessionTeam','gameId','playId','quarter','down','yardsToGo','yardlineSide','Result','gameClock',
                                   'yardlineNumber','kickReturnYardage','playResult','preSnapHomeScore','preSnapVisitorScore','event','homeTeamAbbr','visitorTeamAbbr',
                                   'week'],as_index=True).size()
TENKickoffReturn2018Above25YardsFrame = pd.DataFrame(TENKickoffReturn2018Above25YardsSeries)

mergedTENKickoffReturn2018Above25YardsSeries = mergedTENKickoffReturn2018Above25Yards.groupby(
    ['season','kickoffReturnFormation','tackler','gameId','playId','returnDirectionIntended',
     'returnDirectionActual','kickReturnYardage','homeTeamAbbr','visitorTeamAbbr','week'],as_index=True).size()
mergedTENKickoffReturn2018Above25YardsFrame = pd.DataFrame(mergedTENKickoffReturn2018Above25YardsSeries)

# Above 25 Yards 2019
TENKickoffReturn2019Above25YardsSeries = TENKickoffReturn2019Above25Yards.groupby(['season','possessionTeam','gameId','playId','quarter','down','yardsToGo','yardlineSide','Result','gameClock',
                                   'yardlineNumber','kickReturnYardage','playResult','preSnapHomeScore','preSnapVisitorScore','event','homeTeamAbbr','visitorTeamAbbr',
                                   'week'],as_index=True).size()
TENKickoffReturn2019Above25YardsFrame = pd.DataFrame(TENKickoffReturn2019Above25YardsSeries)

mergedTENKickoffReturn2019Above25YardsSeries = mergedTENKickoffReturn2019Above25Yards.groupby(
    ['season','kickoffReturnFormation','tackler','gameId','playId','returnDirectionIntended',
     'returnDirectionActual','kickReturnYardage','homeTeamAbbr','visitorTeamAbbr','week'],as_index=True).size()
mergedTENKickoffReturn2019Above25YardsFrame = pd.DataFrame(mergedTENKickoffReturn2019Above25YardsSeries)

# Above 25 Yards 2020
TENKickoffReturn2020Above25YardsSeries = TENKickoffReturn2020Above25Yards.groupby(['season','possessionTeam','gameId','playId','quarter','down','yardsToGo','yardlineSide','Result','gameClock',
                                   'yardlineNumber','kickReturnYardage','playResult','preSnapHomeScore','preSnapVisitorScore','event','homeTeamAbbr','visitorTeamAbbr',
                                   'week'],as_index=True).size()
TENKickoffReturn2020Above25YardsFrame = pd.DataFrame(TENKickoffReturn2020Above25YardsSeries)

mergedTENKickoffReturn2020Above25YardsSeries = mergedTENKickoffReturn2020Above25Yards.groupby(
    ['season','kickoffReturnFormation','tackler','gameId','playId','returnDirectionIntended',
     'returnDirectionActual','kickReturnYardage','homeTeamAbbr','visitorTeamAbbr','week'],as_index=True).size()
mergedTENKickoffReturn2020Above25YardsFrame = pd.DataFrame(mergedTENKickoffReturn2020Above25YardsSeries)

# Below 25 Yards 2018
TENKickoffReturn2018Below25YardsSeries = TENKickoffReturn2018Below25Yards.groupby(['season','possessionTeam','gameId','playId','quarter','down','yardsToGo','yardlineSide','Result','gameClock',
                                   'yardlineNumber','kickReturnYardage','playResult','preSnapHomeScore','preSnapVisitorScore','event','homeTeamAbbr','visitorTeamAbbr',
                                   'week'],as_index=True).size()
TENKickoffReturn2018Below25YardsFrame = pd.DataFrame(TENKickoffReturn2018Below25YardsSeries)

mergedTENKickoffReturn2018Below25YardsSeries = mergedTENKickoffReturn2018Below25Yards.groupby(
    ['season','kickoffReturnFormation','tackler','gameId','playId','returnDirectionIntended',
     'returnDirectionActual','kickReturnYardage','homeTeamAbbr','visitorTeamAbbr','week'],as_index=True).size()
mergedTENKickoffReturn2018Below25YardsFrame = pd.DataFrame(mergedTENKickoffReturn2018Below25YardsSeries)

# Below 25 Yards 2019
TENKickoffReturn2019Below25YardsSeries = TENKickoffReturn2019Below25Yards.groupby(['season','possessionTeam','gameId','playId','quarter','down','yardsToGo','yardlineSide','Result','gameClock',
                                   'yardlineNumber','kickReturnYardage','playResult','preSnapHomeScore','preSnapVisitorScore','event','homeTeamAbbr','visitorTeamAbbr',
                                   'week'],as_index=True).size()
TENKickoffReturn2019Below25YardsFrame = pd.DataFrame(TENKickoffReturn2019Below25YardsSeries)

mergedTENKickoffReturn2019Below25YardsSeries = mergedTENKickoffReturn2019Below25Yards.groupby(
    ['season','kickoffReturnFormation','tackler','gameId','playId','returnDirectionIntended',
     'returnDirectionActual','kickReturnYardage','homeTeamAbbr','visitorTeamAbbr','week'],as_index=True).size()
mergedTENKickoffReturn2019Below25YardsFrame = pd.DataFrame(mergedTENKickoffReturn2019Below25YardsSeries)

# Below 25 Yards 2020
TENKickoffReturn2020Below25YardsSeries = TENKickoffReturn2020Below25Yards.groupby(['season','possessionTeam','gameId','playId','quarter','down','yardsToGo','yardlineSide','Result','gameClock',
                                   'yardlineNumber','kickReturnYardage','playResult','preSnapHomeScore','preSnapVisitorScore','event','homeTeamAbbr','visitorTeamAbbr',
                                   'week'],as_index=True).size()
TENKickoffReturn2020Below25YardsFrame = pd.DataFrame(TENKickoffReturn2020Below25YardsSeries)

mergedTENKickoffReturn2020Below25YardsSeries = mergedTENKickoffReturn2020Below25Yards.groupby(
    ['season','kickoffReturnFormation','tackler','gameId','playId','returnDirectionIntended',
     'returnDirectionActual','kickReturnYardage','homeTeamAbbr','visitorTeamAbbr','week'],as_index=True).size()
mergedTENKickoffReturn2020Below25YardsFrame = pd.DataFrame(mergedTENKickoffReturn2020Below25YardsSeries)

# Loss Yards 2018
TENKickoffReturn2018LossYardsSeries = TENKickoffReturn2018LossYards.groupby(['season','possessionTeam','gameId','playId','quarter','down','yardsToGo','yardlineSide','Result','gameClock',
                                   'yardlineNumber','kickReturnYardage','playResult','preSnapHomeScore','preSnapVisitorScore','event','homeTeamAbbr','visitorTeamAbbr',
                                   'week'],as_index=True).size()
TENKickoffReturn2018LossYardsFrame = pd.DataFrame(TENKickoffReturn2018LossYardsSeries)

mergedTENKickoffReturn2018LossYardsSeires = mergedTENKickoffReturn2018LossYards.groupby(
    ['season','kickoffReturnFormation','specialTeamsSafeties','gameId','playId','returnDirectionIntended',
     'returnDirectionActual','kickReturnYardage','homeTeamAbbr','visitorTeamAbbr','week'],as_index=True).size()
mergedTENKickoffReturn2018LossYardsFrame = pd.DataFrame(mergedTENKickoffReturn2018LossYardsSeires)
```
