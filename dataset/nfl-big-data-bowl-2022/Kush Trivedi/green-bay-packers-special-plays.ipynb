**<p id='1'><li>Function to Load Video File</li></p>**

```python
def playVideo(videoFile):
    html = ''
    video = open(videoFile,'rb').read()
    src = 'data:video/mp4;base64,' + b64encode(video).decode()
    html += '<video width=1100 controls autoplay loop><source src="%s" type="video/mp4"></video>' % src 
    return HTML(html)
```

# Data Loading

```python
extraPointsGB2018 = pd.read_csv('../input/green-bay-packers-special-play-2018-to-2020/GB_ExtraPoint_2018.csv')
extraPointsGB2019 = pd.read_csv('../input/green-bay-packers-special-play-2018-to-2020/GB_ExtraPoint_2019.csv')
extraPointsGB2020 = pd.read_csv('../input/green-bay-packers-special-play-2018-to-2020/GB_ExtraPoint_2020.csv')

fieldGoalsGB2018 = pd.read_csv('../input/green-bay-packers-special-play-2018-to-2020/GB_FiledGoals_2018.csv')
fieldGoalsGB2019 = pd.read_csv('../input/green-bay-packers-special-play-2018-to-2020/GB_FiledGoals_2019.csv')
fieldGoalsGB2020 = pd.read_csv('../input/green-bay-packers-special-play-2018-to-2020/GB_FiledGoals_2020.csv')

kickoffsGB2018 = pd.read_csv('../input/green-bay-packers-special-play-2018-to-2020/GB_Kickoffs_2018.csv')
kickoffsGB2019 = pd.read_csv('../input/green-bay-packers-special-play-2018-to-2020/GB_Kickoffs_2019.csv')
kickoffsGB2020 = pd.read_csv('../input/green-bay-packers-special-play-2018-to-2020/GB_Kickoffs_2020.csv')

puntsGB2018 = pd.read_csv('../input/green-bay-packers-special-play-2018-to-2020/GB_Punts_2018.csv')
puntsGB2019 = pd.read_csv('../input/green-bay-packers-special-play-2018-to-2020/GB_Punts_2019.csv')
puntsGB2020 = pd.read_csv('../input/green-bay-packers-special-play-2018-to-2020/GB_Punts_2020.csv')

pFFData = pd.read_csv('../input/nfl-big-data-bowl-2022/PFFScoutingData.csv')
pd.options.display.max_rows = 99999999
```

<a id="3"></a> <br>
# Data Preparation for Kick of Return

```python
GBKickoffReturn2018Above25Yards = kickoffsGB2018[(kickoffsGB2018.Result == 'Kickoff - Return') & (kickoffsGB2018.kickReturnYardage > 25) & (kickoffsGB2018.playResult > 0)]
GBKickoffReturn2019Above25Yards = kickoffsGB2019[(kickoffsGB2019.Result == 'Kickoff - Return') & (kickoffsGB2019.kickReturnYardage > 25) & (kickoffsGB2019.playResult > 0)]
GBKickoffReturn2020Above25Yards = kickoffsGB2020[(kickoffsGB2020.Result == 'Kickoff - Return') & (kickoffsGB2020.kickReturnYardage > 25) & (kickoffsGB2020.playResult > 0)]

GBKickoffReturn2018Below25Yards = kickoffsGB2018[(kickoffsGB2018.Result == 'Kickoff - Return') & (kickoffsGB2018.kickReturnYardage <= 25)]
GBKickoffReturn2019Below25Yards = kickoffsGB2019[(kickoffsGB2019.Result == 'Kickoff - Return') & (kickoffsGB2019.kickReturnYardage <= 25)]
GBKickoffReturn2020Below25Yards = kickoffsGB2020[(kickoffsGB2020.Result == 'Kickoff - Return') & (kickoffsGB2020.kickReturnYardage <= 25)]

GBKickoffReturn2018LossYards = kickoffsGB2018[(kickoffsGB2018.Result == 'Kickoff - Return') & (kickoffsGB2018.playResult < 0)]
GBKickoffReturn2019LossYards = kickoffsGB2019[(kickoffsGB2019.Result == 'Kickoff - Return') & (kickoffsGB2019.playResult < 0)]
GBKickoffReturn2020LossYards = kickoffsGB2020[(kickoffsGB2020.Result == 'Kickoff - Return') & (kickoffsGB2020.playResult < 0)]
```

<a id="4"></a> <br>
# Merging Kickoff Return Data with Pro Football Focus Data

```python
# Merged Data with Kickoff Return Above 25 Yards.
mergedGBKickoffReturn2018Above25Yards = pd.merge(GBKickoffReturn2018Above25Yards, 
                                                 pFFData,  
                                                 how='left', 
                                                 left_on=['gameId','playId'], 
                                                 right_on = ['gameId','playId']
                                                )

mergedGBKickoffReturn2019Above25Yards = pd.merge(GBKickoffReturn2019Above25Yards, 
                                                 pFFData,  
                                                 how='left', 
                                                 left_on=['gameId','playId'], 
                                                 right_on = ['gameId','playId']
                                                )

mergedGBKickoffReturn2020Above25Yards = pd.merge(GBKickoffReturn2020Above25Yards, 
                                                 pFFData,  
                                                 how='left', 
                                                 left_on=['gameId','playId'], 
                                                 right_on = ['gameId','playId']
                                                )

# Merged Data with Kickoff Return Below 25 Yards.
mergedGBKickoffReturn2018Below25Yards = pd.merge(GBKickoffReturn2018Below25Yards, 
                                                 pFFData,  
                                                 how='left', 
                                                 left_on=['gameId','playId'], 
                                                 right_on = ['gameId','playId']
                                                )

mergedGBKickoffReturn2019Below25Yards = pd.merge(GBKickoffReturn2019Below25Yards, 
                                                 pFFData,  
                                                 how='left', 
                                                 left_on=['gameId','playId'], 
                                                 right_on = ['gameId','playId']
                                                )

mergedGBKickoffReturn2020Below25Yards = pd.merge(GBKickoffReturn2020Below25Yards, 
                                                 pFFData,  
                                                 how='left', 
                                                 left_on=['gameId','playId'], 
                                                 right_on = ['gameId','playId']
                                                )

# Merged Data with Kickoff Return Lossing Yards.
mergedGBKickoffReturn2018LossYards = pd.merge(GBKickoffReturn2018LossYards, 
                                                 pFFData,  
                                                 how='left', 
                                                 left_on=['gameId','playId'], 
                                                 right_on = ['gameId','playId']
                                                )

mergedGBKickoffReturn2019LossYards = pd.merge(GBKickoffReturn2019LossYards, 
                                                 pFFData,  
                                                 how='left', 
                                                 left_on=['gameId','playId'], 
                                                 right_on = ['gameId','playId']
                                                )

mergedGBKickoffReturn2020LossYards = pd.merge(GBKickoffReturn2020LossYards, 
                                                 pFFData,  
                                                 how='left', 
                                                 left_on=['gameId','playId'], 
                                                 right_on = ['gameId','playId']
                                                )
```
