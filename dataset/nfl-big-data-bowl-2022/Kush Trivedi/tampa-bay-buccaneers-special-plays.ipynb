**<p id='1'><li>Function to Load Video File</li></p>**

```python
def playVideo(videoFile):
    html = ''
    video = open(videoFile,'rb').read()
    src = 'data:video/mp4;base64,' + b64encode(video).decode()
    html += '<video width=1100 controls autoplay loop><source src="%s" type="video/mp4"></video>' % src 
    return HTML(html)
```

# Data Loading

```python
extraPointsTB2018 = pd.read_csv('../input/tampa-bay-buccaneers-special-play-2018-to-2020/TB_ExtraPoint_2018.csv')
extraPointsTB2019 = pd.read_csv('../input/tampa-bay-buccaneers-special-play-2018-to-2020/TB_ExtraPoint_2019.csv')
extraPointsTB2020 = pd.read_csv('../input/tampa-bay-buccaneers-special-play-2018-to-2020/TB_ExtraPoint_2020.csv')

fieldGoalsTB2018 = pd.read_csv('../input/tampa-bay-buccaneers-special-play-2018-to-2020/TB_FiledGoals_2018.csv')
fieldGoalsTB2019 = pd.read_csv('../input/tampa-bay-buccaneers-special-play-2018-to-2020/TB_FiledGoals_2019.csv')
fieldGoalsTB2020 = pd.read_csv('../input/tampa-bay-buccaneers-special-play-2018-to-2020/TB_FiledGoals_2020.csv')

kickoffsTB2018 = pd.read_csv('../input/tampa-bay-buccaneers-special-play-2018-to-2020/TB_Kickoffs_2018.csv')
kickoffsTB2019 = pd.read_csv('../input/tampa-bay-buccaneers-special-play-2018-to-2020/TB_Kickoffs_2019.csv')
kickoffsTB2020 = pd.read_csv('../input/tampa-bay-buccaneers-special-play-2018-to-2020/TB_Kickoffs_2020.csv')

puntsTB2018 = pd.read_csv('../input/tampa-bay-buccaneers-special-play-2018-to-2020/TB_Punts_2018.csv')
puntsTB2019 = pd.read_csv('../input/tampa-bay-buccaneers-special-play-2018-to-2020/TB_Punts_2019.csv')
puntsTB2020 = pd.read_csv('../input/tampa-bay-buccaneers-special-play-2018-to-2020/TB_Punts_2020.csv')

pFFData = pd.read_csv('../input/nfl-big-data-bowl-2022/PFFScoutingData.csv')
pd.options.display.max_rows = 99999999
```

<a id="3"></a> <br>
# Data Preparation for Kick of Return

```python
TBKickoffReturn2018Above25Yards = kickoffsTB2018[(kickoffsTB2018.Result == 'Kickoff - Return') & (kickoffsTB2018.kickReturnYardage > 25)]
TBKickoffReturn2019Above25Yards = kickoffsTB2019[(kickoffsTB2019.Result == 'Kickoff - Return') & (kickoffsTB2019.kickReturnYardage > 25)]
TBKickoffReturn2020Above25Yards = kickoffsTB2020[(kickoffsTB2020.Result == 'Kickoff - Return') & (kickoffsTB2020.kickReturnYardage > 25) & (kickoffsTB2020.playResult > 0)]

TBKickoffReturn2018Below25Yards = kickoffsTB2018[(kickoffsTB2018.Result == 'Kickoff - Return') & (kickoffsTB2018.kickReturnYardage <= 25)]
TBKickoffReturn2019Below25Yards = kickoffsTB2019[(kickoffsTB2019.Result == 'Kickoff - Return') & (kickoffsTB2019.kickReturnYardage <= 25)]
TBKickoffReturn2020Below25Yards = kickoffsTB2020[(kickoffsTB2020.Result == 'Kickoff - Return') & (kickoffsTB2020.kickReturnYardage <= 25)]

TBKickoffReturn2020LossYards = kickoffsTB2020[(kickoffsTB2020.Result == 'Kickoff - Return') & (kickoffsTB2020.playResult < 0)]
```

<a id="4"></a> <br>
# Merging Kickoff Return Data with Pro Football Focus Data

```python
# Merged Data with Kickoff Return Above 25 Yards.
mergedTBKickoffReturn2018Above25Yards = pd.merge(TBKickoffReturn2018Above25Yards, 
                                                 pFFData,  
                                                 how='left', 
                                                 left_on=['gameId','playId'], 
                                                 right_on = ['gameId','playId']
                                                )

mergedTBKickoffReturn2019Above25Yards = pd.merge(TBKickoffReturn2019Above25Yards, 
                                                 pFFData,  
                                                 how='left', 
                                                 left_on=['gameId','playId'], 
                                                 right_on = ['gameId','playId']
                                                )

mergedTBKickoffReturn2020Above25Yards = pd.merge(TBKickoffReturn2020Above25Yards, 
                                                 pFFData,  
                                                 how='left', 
                                                 left_on=['gameId','playId'], 
                                                 right_on = ['gameId','playId']
                                                )

# Merged Data with Kickoff Return Below 25 Yards.
mergedTBKickoffReturn2018Below25Yards = pd.merge(TBKickoffReturn2018Below25Yards, 
                                                 pFFData,  
                                                 how='left', 
                                                 left_on=['gameId','playId'], 
                                                 right_on = ['gameId','playId']
                                                )

mergedTBKickoffReturn2019Below25Yards = pd.merge(TBKickoffReturn2019Below25Yards, 
                                                 pFFData,  
                                                 how='left', 
                                                 left_on=['gameId','playId'], 
                                                 right_on = ['gameId','playId']
                                                )

mergedTBKickoffReturn2020Below25Yards = pd.merge(TBKickoffReturn2020Below25Yards, 
                                                 pFFData,  
                                                 how='left', 
                                                 left_on=['gameId','playId'], 
                                                 right_on = ['gameId','playId']
                                                )

# Merged Data with Kickoff Return Lossing Yards.
mergedTBKickoffReturn2020LossYards = pd.merge(TBKickoffReturn2020LossYards, 
                                                 pFFData,  
                                                 how='left', 
                                                 left_on=['gameId','playId'], 
                                                 right_on = ['gameId','playId']
                                                )
```

<a id="5"></a> <br>
# Grouping DataFrames

```python
# Above 25 Yards 2018
TBKickoffReturn2018Above25YardsSeries = TBKickoffReturn2018Above25Yards.groupby(['season','possessionTeam','gameId','playId','quarter','down','yardsToGo','yardlineSide','Result','gameClock',
                                   'yardlineNumber','kickReturnYardage','playResult','preSnapHomeScore','preSnapVisitorScore','event','homeTeamAbbr','visitorTeamAbbr',
                                   'week'],as_index=True).size()
TBKickoffReturn2018Above25YardsFrame = pd.DataFrame(TBKickoffReturn2018Above25YardsSeries)

mergedTBKickoffReturn2018Above25YardsSeries = mergedTBKickoffReturn2018Above25Yards.groupby(
    ['season','kickoffReturnFormation','tackler','gameId','playId','returnDirectionIntended',
     'returnDirectionActual','kickReturnYardage','homeTeamAbbr','visitorTeamAbbr','week'],as_index=True).size()
mergedTBKickoffReturn2018Above25YardsFrame = pd.DataFrame(mergedTBKickoffReturn2018Above25YardsSeries)

# Above 25 Yards 2019
TBKickoffReturn2019Above25YardsSeries = TBKickoffReturn2019Above25Yards.groupby(['season','possessionTeam','gameId','playId','quarter','down','yardsToGo','yardlineSide','Result','gameClock',
                                   'yardlineNumber','kickReturnYardage','playResult','preSnapHomeScore','preSnapVisitorScore','event','homeTeamAbbr','visitorTeamAbbr',
                                   'week'],as_index=True).size()
TBKickoffReturn2019Above25YardsFrame = pd.DataFrame(TBKickoffReturn2019Above25YardsSeries)

mergedTBKickoffReturn2019Above25YardsSeries = mergedTBKickoffReturn2019Above25Yards.groupby(
    ['season','kickoffReturnFormation','tackler','gameId','playId','returnDirectionIntended',
     'returnDirectionActual','kickReturnYardage','homeTeamAbbr','visitorTeamAbbr','week'],as_index=True).size()
mergedTBKickoffReturn2019Above25YardsFrame = pd.DataFrame(mergedTBKickoffReturn2019Above25YardsSeries)

# Above 25 Yards 2020
TBKickoffReturn2020Above25YardsSeries = TBKickoffReturn2020Above25Yards.groupby(['season','possessionTeam','gameId','playId','quarter','down','yardsToGo','yardlineSide','Result','gameClock',
                                   'yardlineNumber','kickReturnYardage','playResult','preSnapHomeScore','preSnapVisitorScore','event','homeTeamAbbr','visitorTeamAbbr',
                                   'week'],as_index=True).size()
TBKickoffReturn2020Above25YardsFrame = pd.DataFrame(TBKickoffReturn2020Above25YardsSeries)

mergedTBKickoffReturn2020Above25YardsSeries = mergedTBKickoffReturn2020Above25Yards.groupby(
    ['season','kickoffReturnFormation','tackler','gameId','playId','returnDirectionIntended',
     'returnDirectionActual','kickReturnYardage','homeTeamAbbr','visitorTeamAbbr','week'],as_index=True).size()
mergedTBKickoffReturn2020Above25YardsFrame = pd.DataFrame(mergedTBKickoffReturn2020Above25YardsSeries)

# Below 25 Yards 2018
TBKickoffReturn2018Below25YardsSeries = TBKickoffReturn2018Below25Yards.groupby(['season','possessionTeam','gameId','playId','quarter','down','yardsToGo','yardlineSide','Result','gameClock',
                                   'yardlineNumber','kickReturnYardage','playResult','preSnapHomeScore','preSnapVisitorScore','event','homeTeamAbbr','visitorTeamAbbr',
                                   'week'],as_index=True).size()
TBKickoffReturn2018Below25YardsFrame = pd.DataFrame(TBKickoffReturn2018Below25YardsSeries)

mergedTBKickoffReturn2018Below25YardsSeries = mergedTBKickoffReturn2018Below25Yards.groupby(
    ['season','kickoffReturnFormation','tackler','gameId','playId','returnDirectionIntended',
     'returnDirectionActual','kickReturnYardage','homeTeamAbbr','visitorTeamAbbr','week'],as_index=True).size()
mergedTBKickoffReturn2018Below25YardsFrame = pd.DataFrame(mergedTBKickoffReturn2018Below25YardsSeries)

# Below 25 Yards 2018
TBKickoffReturn2019Below25YardsSeries = TBKickoffReturn2019Below25Yards.groupby(['season','possessionTeam','gameId','playId','quarter','down','yardsToGo','yardlineSide','Result','gameClock',
                                   'yardlineNumber','kickReturnYardage','playResult','preSnapHomeScore','preSnapVisitorScore','event','homeTeamAbbr','visitorTeamAbbr',
                                   'week'],as_index=True).size()
TBKickoffReturn2019Below25YardsFrame = pd.DataFrame(TBKickoffReturn2019Below25YardsSeries)

mergedTBKickoffReturn2019Below25YardsSeries = mergedTBKickoffReturn2019Below25Yards.groupby(
    ['season','kickoffReturnFormation','tackler','gameId','playId','returnDirectionIntended',
     'returnDirectionActual','kickReturnYardage','homeTeamAbbr','visitorTeamAbbr','week'],as_index=True).size()
mergedTBKickoffReturn2019Below25YardsFrame = pd.DataFrame(mergedTBKickoffReturn2019Below25YardsSeries)

# Below 25 Yards 2018
TBKickoffReturn2020Below25YardsSeries = TBKickoffReturn2020Below25Yards.groupby(['season','possessionTeam','gameId','playId','quarter','down','yardsToGo','yardlineSide','Result','gameClock',
                                   'yardlineNumber','kickReturnYardage','playResult','preSnapHomeScore','preSnapVisitorScore','event','homeTeamAbbr','visitorTeamAbbr',
                                   'week'],as_index=True).size()
TBKickoffReturn2020Below25YardsFrame = pd.DataFrame(TBKickoffReturn2020Below25YardsSeries)

mergedTBKickoffReturn2020Below25YardsSeries = mergedTBKickoffReturn2020Below25Yards.groupby(
    ['season','kickoffReturnFormation','tackler','gameId','playId','returnDirectionIntended',
     'returnDirectionActual','kickReturnYardage','homeTeamAbbr','visitorTeamAbbr','week'],as_index=True).size()
mergedTBKickoffReturn2020Below25YardsFrame = pd.DataFrame(mergedTBKickoffReturn2020Below25YardsSeries)

# Loosing Yards 2020
TBKickoffReturn2020LossYardsSeries = TBKickoffReturn2020LossYards.groupby(['season','possessionTeam','gameId','playId','quarter','down','yardsToGo','yardlineSide','Result','gameClock',
                                   'yardlineNumber','kickReturnYardage','playResult','preSnapHomeScore','preSnapVisitorScore','event','homeTeamAbbr','visitorTeamAbbr',
                                   'week'],as_index=True).size()
TBKickoffReturn2020LossYardsFrame = pd.DataFrame(TBKickoffReturn2020LossYardsSeries)

mergedTBKickoffReturn2020LossYardsSeires = mergedTBKickoffReturn2020LossYards.groupby(
    ['season','kickoffReturnFormation','specialTeamsSafeties','gameId','playId','returnDirectionIntended',
     'returnDirectionActual','kickReturnYardage','homeTeamAbbr','visitorTeamAbbr','week'],as_index=True).size()
mergedTBKickoffReturn2020LossYardsFrame = pd.DataFrame(mergedTBKickoffReturn2020LossYardsSeires)
```
