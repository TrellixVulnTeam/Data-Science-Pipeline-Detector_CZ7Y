**<p id='1'><li>Function to Load Video File</li></p>**

```python
def playVideo(videoFile):
    html = ''
    video = open(videoFile,'rb').read()
    src = 'data:video/mp4;base64,' + b64encode(video).decode()
    html += '<video width=1100 controls autoplay loop><source src="%s" type="video/mp4"></video>' % src 
    return HTML(html)
```

# Data Loading

```python
extraPointsDAL2018 = pd.read_csv('../input/dallas-cowboys-special-play-2018-to-2020/DAL_ExtraPoint_2018.csv')
extraPointsDAL2019 = pd.read_csv('../input/dallas-cowboys-special-play-2018-to-2020/DAL_ExtraPoint_2019.csv')
extraPointsDAL2020 = pd.read_csv('../input/dallas-cowboys-special-play-2018-to-2020/DAL_ExtraPoint_2020.csv')

fieldGoalsDAL2018 = pd.read_csv('../input/dallas-cowboys-special-play-2018-to-2020/DAL_FiledGoals_2018.csv')
fieldGoalsDAL2019 = pd.read_csv('../input/dallas-cowboys-special-play-2018-to-2020/DAL_FiledGoals_2019.csv')
fieldGoalsDAL2020 = pd.read_csv('../input/dallas-cowboys-special-play-2018-to-2020/DAL_FiledGoals_2020.csv')

kickoffsDAL2018 = pd.read_csv('../input/dallas-cowboys-special-play-2018-to-2020/DAL_Kickoffs_2018.csv')
kickoffsDAL2019 = pd.read_csv('../input/dallas-cowboys-special-play-2018-to-2020/DAL_Kickoffs_2019.csv')
kickoffsDAL2020 = pd.read_csv('../input/dallas-cowboys-special-play-2018-to-2020/DAL_Kickoffs_2020.csv')

puntsDAL2018 = pd.read_csv('../input/dallas-cowboys-special-play-2018-to-2020/DAL_Punts_2018.csv')
puntsDAL2019 = pd.read_csv('../input/dallas-cowboys-special-play-2018-to-2020/DAL_Punts_2019.csv')
puntsDAL2020 = pd.read_csv('../input/dallas-cowboys-special-play-2018-to-2020/DAL_Punts_2020.csv')

pFFData = pd.read_csv('../input/nfl-big-data-bowl-2022/PFFScoutingData.csv')
pd.options.display.max_rows = 99999999
```

<a id="3"></a> <br>
# Data Preparation for Kick of Return

```python
DALKickoffReturn2018Above25Yards = kickoffsDAL2018[(kickoffsDAL2018.Result == 'Kickoff - Return') & (kickoffsDAL2018.kickReturnYardage > 25)]
DALKickoffReturn2019Above25Yards = kickoffsDAL2019[(kickoffsDAL2019.Result == 'Kickoff - Return') & (kickoffsDAL2019.kickReturnYardage > 25)]
DALKickoffReturn2020Above25Yards = kickoffsDAL2020[(kickoffsDAL2020.Result == 'Kickoff - Return') & (kickoffsDAL2020.kickReturnYardage > 25)]

DALKickoffReturn2018Below25Yards = kickoffsDAL2018[(kickoffsDAL2018.Result == 'Kickoff - Return') & (kickoffsDAL2018.kickReturnYardage <= 25)]
DALKickoffReturn2019Below25Yards = kickoffsDAL2019[(kickoffsDAL2019.Result == 'Kickoff - Return') & (kickoffsDAL2019.kickReturnYardage <= 25)]
DALKickoffReturn2020Below25Yards = kickoffsDAL2020[(kickoffsDAL2020.Result == 'Kickoff - Return') & (kickoffsDAL2020.kickReturnYardage <= 25)]
```

<a id="4"></a> <br>
# Merging Kickoff Return Data with Pro Football Focus Data

```python
# Merged Data with Kickoff Return Above 25 Yards.
mergedDALKickoffReturn2018Above25Yards = pd.merge(DALKickoffReturn2018Above25Yards, 
                                                 pFFData,  
                                                 how='left', 
                                                 left_on=['gameId','playId'], 
                                                 right_on = ['gameId','playId']
                                                )

mergedDALKickoffReturn2019Above25Yards = pd.merge(DALKickoffReturn2019Above25Yards, 
                                                 pFFData,  
                                                 how='left', 
                                                 left_on=['gameId','playId'], 
                                                 right_on = ['gameId','playId']
                                                )

mergedDALKickoffReturn2020Above25Yards = pd.merge(DALKickoffReturn2020Above25Yards, 
                                                 pFFData,  
                                                 how='left', 
                                                 left_on=['gameId','playId'], 
                                                 right_on = ['gameId','playId']
                                                )

# Merged Data with Kickoff Return Below 25 Yards.
mergedDALKickoffReturn2018Below25Yards = pd.merge(DALKickoffReturn2018Below25Yards, 
                                                 pFFData,  
                                                 how='left', 
                                                 left_on=['gameId','playId'], 
                                                 right_on = ['gameId','playId']
                                                )

mergedDALKickoffReturn2019Below25Yards = pd.merge(DALKickoffReturn2019Below25Yards, 
                                                 pFFData,  
                                                 how='left', 
                                                 left_on=['gameId','playId'], 
                                                 right_on = ['gameId','playId']
                                                )

mergedDALKickoffReturn2020Below25Yards = pd.merge(DALKickoffReturn2020Below25Yards, 
                                                 pFFData,  
                                                 how='left', 
                                                 left_on=['gameId','playId'], 
                                                 right_on = ['gameId','playId']
                                                )
```

<a id="5"></a> <br>

```python
# Above 25 Yards 2018
DALKickoffReturn2018Above25YardsSeries = DALKickoffReturn2018Above25Yards.groupby(['season','possessionTeam','gameId','playId','quarter','down','yardsToGo','yardlineSide','Result','gameClock',
                                   'yardlineNumber','kickReturnYardage','playResult','preSnapHomeScore','preSnapVisitorScore','event','homeTeamAbbr','visitorTeamAbbr',
                                   'week'],as_index=True).size()
DALKickoffReturn2018Above25YardsFrame = pd.DataFrame(DALKickoffReturn2018Above25YardsSeries)

mergedDALKickoffReturn2018Above25YardsSeries = mergedDALKickoffReturn2018Above25Yards.groupby(
    ['season','kickoffReturnFormation','tackler','gameId','playId','returnDirectionIntended',
     'returnDirectionActual','kickReturnYardage','homeTeamAbbr','visitorTeamAbbr','week'],as_index=True).size()
mergedDALKickoffReturn2018Above25YardsFrame = pd.DataFrame(mergedDALKickoffReturn2018Above25YardsSeries)

# Above 25 Yards 2019
DALKickoffReturn2019Above25YardsSeries = DALKickoffReturn2019Above25Yards.groupby(['season','possessionTeam','gameId','playId','quarter','down','yardsToGo','yardlineSide','Result','gameClock',
                                   'yardlineNumber','kickReturnYardage','playResult','preSnapHomeScore','preSnapVisitorScore','event','homeTeamAbbr','visitorTeamAbbr',
                                   'week'],as_index=True).size()
DALKickoffReturn2019Above25YardsFrame = pd.DataFrame(DALKickoffReturn2019Above25YardsSeries)

mergedDALKickoffReturn2019Above25YardsSeries = mergedDALKickoffReturn2019Above25Yards.groupby(
    ['season','kickoffReturnFormation','tackler','gameId','playId','returnDirectionIntended',
     'returnDirectionActual','kickReturnYardage','homeTeamAbbr','visitorTeamAbbr','week'],as_index=True).size()
mergedDALKickoffReturn2019Above25YardsFrame = pd.DataFrame(mergedDALKickoffReturn2019Above25YardsSeries)

# Above 25 Yards 2020
DALKickoffReturn2020Above25YardsSeries = DALKickoffReturn2020Above25Yards.groupby(['season','possessionTeam','gameId','playId','quarter','down','yardsToGo','yardlineSide','Result','gameClock',
                                   'yardlineNumber','kickReturnYardage','playResult','preSnapHomeScore','preSnapVisitorScore','event','homeTeamAbbr','visitorTeamAbbr',
                                   'week'],as_index=True).size()
DALKickoffReturn2020Above25YardsFrame = pd.DataFrame(DALKickoffReturn2020Above25YardsSeries)

mergedDALKickoffReturn2020Above25YardsSeries = mergedDALKickoffReturn2020Above25Yards.groupby(
    ['season','kickoffReturnFormation','tackler','gameId','playId','returnDirectionIntended',
     'returnDirectionActual','kickReturnYardage','homeTeamAbbr','visitorTeamAbbr','week'],as_index=True).size()
mergedDALKickoffReturn2020Above25YardsFrame = pd.DataFrame(mergedDALKickoffReturn2020Above25YardsSeries)

# Below 25 Yards 2018
DALKickoffReturn2018Below25YardsSeries = DALKickoffReturn2018Below25Yards.groupby(['season','possessionTeam','gameId','playId','quarter','down','yardsToGo','yardlineSide','Result','gameClock',
                                   'yardlineNumber','kickReturnYardage','playResult','preSnapHomeScore','preSnapVisitorScore','event','homeTeamAbbr','visitorTeamAbbr',
                                   'week'],as_index=True).size()
DALKickoffReturn2018Below25YardsFrame = pd.DataFrame(DALKickoffReturn2018Below25YardsSeries)

mergedDALKickoffReturn2018Below25YardsSeries = mergedDALKickoffReturn2018Below25Yards.groupby(
    ['season','kickoffReturnFormation','tackler','gameId','playId','returnDirectionIntended',
     'returnDirectionActual','kickReturnYardage','homeTeamAbbr','visitorTeamAbbr','week'],as_index=True).size()
mergedDALKickoffReturn2018Below25YardsFrame = pd.DataFrame(mergedDALKickoffReturn2018Below25YardsSeries)

# Below 25 Yards 2019
DALKickoffReturn2019Below25YardsSeries = DALKickoffReturn2019Below25Yards.groupby(['season','possessionTeam','gameId','playId','quarter','down','yardsToGo','yardlineSide','Result','gameClock',
                                   'yardlineNumber','kickReturnYardage','playResult','preSnapHomeScore','preSnapVisitorScore','event','homeTeamAbbr','visitorTeamAbbr',
                                   'week'],as_index=True).size()
DALKickoffReturn2019Below25YardsFrame = pd.DataFrame(DALKickoffReturn2019Below25YardsSeries)

mergedDALKickoffReturn2019Below25YardsSeries = mergedDALKickoffReturn2019Below25Yards.groupby(
    ['season','kickoffReturnFormation','tackler','gameId','playId','returnDirectionIntended',
     'returnDirectionActual','kickReturnYardage','homeTeamAbbr','visitorTeamAbbr','week'],as_index=True).size()
mergedDALKickoffReturn2019Below25YardsFrame = pd.DataFrame(mergedDALKickoffReturn2019Below25YardsSeries)

# Below 25 Yards 2020
DALKickoffReturn2020Below25YardsSeries = DALKickoffReturn2020Below25Yards.groupby(['season','possessionTeam','gameId','playId','quarter','down','yardsToGo','yardlineSide','Result','gameClock',
                                   'yardlineNumber','kickReturnYardage','playResult','preSnapHomeScore','preSnapVisitorScore','event','homeTeamAbbr','visitorTeamAbbr',
                                   'week'],as_index=True).size()
DALKickoffReturn2020Below25YardsFrame = pd.DataFrame(DALKickoffReturn2020Below25YardsSeries)

mergedDALKickoffReturn2020Below25YardsSeries = mergedDALKickoffReturn2020Below25Yards.groupby(
    ['season','kickoffReturnFormation','tackler','gameId','playId','returnDirectionIntended',
     'returnDirectionActual','kickReturnYardage','homeTeamAbbr','visitorTeamAbbr','week'],as_index=True).size()
mergedDALKickoffReturn2020Below25YardsFrame = pd.DataFrame(mergedDALKickoffReturn2020Below25YardsSeries)
```
