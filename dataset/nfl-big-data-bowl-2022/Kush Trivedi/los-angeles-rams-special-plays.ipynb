**<p id='1'><li>Function to Load Video File</li></p>**

```python
def playVideo(videoFile):
    html = ''
    video = open(videoFile,'rb').read()
    src = 'data:video/mp4;base64,' + b64encode(video).decode()
    html += '<video width=1100 controls autoplay loop><source src="%s" type="video/mp4"></video>' % src 
    return HTML(html)
```

# Data Loading

```python
extraPointsLA2018 = pd.read_csv('../input/los-angeles-rams-special-play-2018-to-2020/LA_ExtraPoint_2018.csv')
extraPointsLA2019 = pd.read_csv('../input/los-angeles-rams-special-play-2018-to-2020/LA_ExtraPoint_2019.csv')
extraPointsLA2020 = pd.read_csv('../input/los-angeles-rams-special-play-2018-to-2020/LA_ExtraPoint_2020.csv')

fieldGoalsLA2018 = pd.read_csv('../input/los-angeles-rams-special-play-2018-to-2020/LA_FiledGoals_2018.csv')
fieldGoalsLA2019 = pd.read_csv('../input/los-angeles-rams-special-play-2018-to-2020/LA_FiledGoals_2019.csv')
fieldGoalsLA2020 = pd.read_csv('../input/los-angeles-rams-special-play-2018-to-2020/LA_FiledGoals_2020.csv')

kickoffsLA2018 = pd.read_csv('../input/los-angeles-rams-special-play-2018-to-2020/LA_Kickoffs_2018.csv')
kickoffsLA2019 = pd.read_csv('../input/los-angeles-rams-special-play-2018-to-2020/LA_Kickoffs_2019.csv')
kickoffsLA2020 = pd.read_csv('../input/los-angeles-rams-special-play-2018-to-2020/LA_Kickoffs_2020.csv')

puntsLA2018 = pd.read_csv('../input/los-angeles-rams-special-play-2018-to-2020/LA_Punts_2018.csv')
puntsLA2019 = pd.read_csv('../input/los-angeles-rams-special-play-2018-to-2020/LA_Punts_2019.csv')
puntsLA2020 = pd.read_csv('../input/los-angeles-rams-special-play-2018-to-2020/LA_Punts_2020.csv')

pFFData = pd.read_csv('../input/nfl-big-data-bowl-2022/PFFScoutingData.csv')
pd.options.display.max_rows = 99999999
```

<a id="3"></a> <br>
# Data Preparation for Kick of Return

```python
LAKickoffReturn2018Above25Yards = kickoffsLA2018[(kickoffsLA2018.Result == 'Kickoff - Return') & (kickoffsLA2018.kickReturnYardage > 25)]
LAKickoffReturn2019Above25Yards = kickoffsLA2019[(kickoffsLA2019.Result == 'Kickoff - Return') & (kickoffsLA2019.kickReturnYardage > 25) & (kickoffsLA2019.playResult > 0)]
LAKickoffReturn2020Above25Yards = kickoffsLA2020[(kickoffsLA2020.Result == 'Kickoff - Return') & (kickoffsLA2020.kickReturnYardage > 25)]

LAKickoffReturn2018Below25Yards = kickoffsLA2018[(kickoffsLA2018.Result == 'Kickoff - Return') & (kickoffsLA2018.kickReturnYardage <= 25)]
LAKickoffReturn2019Below25Yards = kickoffsLA2019[(kickoffsLA2019.Result == 'Kickoff - Return') & (kickoffsLA2019.kickReturnYardage <= 25)]
LAKickoffReturn2020Below25Yards = kickoffsLA2020[(kickoffsLA2020.Result == 'Kickoff - Return') & (kickoffsLA2020.kickReturnYardage <= 25)]

LAKickoffReturn2019LossYards = kickoffsLA2019[(kickoffsLA2019.Result == 'Kickoff - Return') & (kickoffsLA2019.playResult < 0)]
```

<a id="4"></a> <br>
# Merging Kickoff Return Data with Pro Football Focus Data

```python
# Merged Data with Kickoff Return Above 25 Yards.
mergedLAKickoffReturn2018Above25Yards = pd.merge(LAKickoffReturn2018Above25Yards, 
                                                 pFFData,  
                                                 how='left', 
                                                 left_on=['gameId','playId'], 
                                                 right_on = ['gameId','playId']
                                                )

mergedLAKickoffReturn2019Above25Yards = pd.merge(LAKickoffReturn2019Above25Yards, 
                                                 pFFData,  
                                                 how='left', 
                                                 left_on=['gameId','playId'], 
                                                 right_on = ['gameId','playId']
                                                )

mergedLAKickoffReturn2020Above25Yards = pd.merge(LAKickoffReturn2020Above25Yards, 
                                                 pFFData,  
                                                 how='left', 
                                                 left_on=['gameId','playId'], 
                                                 right_on = ['gameId','playId']
                                                )

# Merged Data with Kickoff Return Below 25 Yards.
mergedLAKickoffReturn2018Below25Yards = pd.merge(LAKickoffReturn2018Below25Yards, 
                                                 pFFData,  
                                                 how='left', 
                                                 left_on=['gameId','playId'], 
                                                 right_on = ['gameId','playId']
                                                )

mergedLAKickoffReturn2019Below25Yards = pd.merge(LAKickoffReturn2019Below25Yards, 
                                                 pFFData,  
                                                 how='left', 
                                                 left_on=['gameId','playId'], 
                                                 right_on = ['gameId','playId']
                                                )

mergedLAKickoffReturn2020Below25Yards = pd.merge(LAKickoffReturn2020Below25Yards, 
                                                 pFFData,  
                                                 how='left', 
                                                 left_on=['gameId','playId'], 
                                                 right_on = ['gameId','playId']
                                                )

# Merged Data with Kickoff Return Lossing Yards.
mergedLAKickoffReturn2019LossYards = pd.merge(LAKickoffReturn2019LossYards, 
                                                 pFFData,  
                                                 how='left', 
                                                 left_on=['gameId','playId'], 
                                                 right_on = ['gameId','playId']
                                                )
```

```python
# Above 25 Yards 2018
LAKickoffReturn2018Above25YardsSeries = LAKickoffReturn2018Above25Yards.groupby(['season','possessionTeam','gameId','playId','quarter','down','yardsToGo','yardlineSide','Result','gameClock',
                                   'yardlineNumber','kickReturnYardage','playResult','preSnapHomeScore','preSnapVisitorScore','event','homeTeamAbbr','visitorTeamAbbr',
                                   'week'],as_index=True).size()
LAKickoffReturn2018Above25YardsFrame = pd.DataFrame(LAKickoffReturn2018Above25YardsSeries)

mergedLAKickoffReturn2018Above25YardsSeries = mergedLAKickoffReturn2018Above25Yards.groupby(
    ['season','kickoffReturnFormation','tackler','gameId','playId','returnDirectionIntended',
     'returnDirectionActual','kickReturnYardage','homeTeamAbbr','visitorTeamAbbr','week'],as_index=True).size()
mergedLAKickoffReturn2018Above25YardsFrame = pd.DataFrame(mergedLAKickoffReturn2018Above25YardsSeries)


# Above 25 Yards 2019
LAKickoffReturn2019Above25YardsSeries = LAKickoffReturn2019Above25Yards.groupby(['season','possessionTeam','gameId','playId','quarter','down','yardsToGo','yardlineSide','Result','gameClock',
                                   'yardlineNumber','kickReturnYardage','playResult','preSnapHomeScore','preSnapVisitorScore','event','homeTeamAbbr','visitorTeamAbbr',
                                   'week'],as_index=True).size()
LAKickoffReturn2019Above25YardsFrame = pd.DataFrame(LAKickoffReturn2019Above25YardsSeries)

mergedLAKickoffReturn2019Above25YardsSeries = mergedLAKickoffReturn2019Above25Yards.groupby(
    ['season','kickoffReturnFormation','tackler','gameId','playId','returnDirectionIntended',
     'returnDirectionActual','kickReturnYardage','homeTeamAbbr','visitorTeamAbbr','week'],as_index=True).size()
mergedLAKickoffReturn2019Above25YardsFrame = pd.DataFrame(mergedLAKickoffReturn2019Above25YardsSeries)

# Above 25 Yards 2020
LAKickoffReturn2020Above25YardsSeries = LAKickoffReturn2020Above25Yards.groupby(['season','possessionTeam','gameId','playId','quarter','down','yardsToGo','yardlineSide','Result','gameClock',
                                   'yardlineNumber','kickReturnYardage','playResult','preSnapHomeScore','preSnapVisitorScore','event','homeTeamAbbr','visitorTeamAbbr',
                                   'week'],as_index=True).size()
LAKickoffReturn2020Above25YardsFrame = pd.DataFrame(LAKickoffReturn2020Above25YardsSeries)

mergedLAKickoffReturn2020Above25YardsSeries = mergedLAKickoffReturn2020Above25Yards.groupby(
    ['season','kickoffReturnFormation','tackler','gameId','playId','returnDirectionIntended',
     'returnDirectionActual','kickReturnYardage','homeTeamAbbr','visitorTeamAbbr','week'],as_index=True).size()
mergedLAKickoffReturn2020Above25YardsFrame = pd.DataFrame(mergedLAKickoffReturn2020Above25YardsSeries)

# Below 25 Yards 2018
LAKickoffReturn2018Below25YardsSeries = LAKickoffReturn2018Below25Yards.groupby(['season','possessionTeam','gameId','playId','quarter','down','yardsToGo','yardlineSide','Result','gameClock',
                                   'yardlineNumber','kickReturnYardage','playResult','preSnapHomeScore','preSnapVisitorScore','event','homeTeamAbbr','visitorTeamAbbr',
                                   'week'],as_index=True).size()
LAKickoffReturn2018Below25YardsFrame = pd.DataFrame(LAKickoffReturn2018Below25YardsSeries)

mergedLAKickoffReturn2018Below25YardsSeries = mergedLAKickoffReturn2018Below25Yards.groupby(
    ['season','kickoffReturnFormation','tackler','gameId','playId','returnDirectionIntended',
     'returnDirectionActual','kickReturnYardage','homeTeamAbbr','visitorTeamAbbr','week'],as_index=True).size()
mergedLAKickoffReturn2018Below25YardsFrame = pd.DataFrame(mergedLAKickoffReturn2018Below25YardsSeries)

# Below 25 Yards 2019
LAKickoffReturn2019Below25YardsSeries = LAKickoffReturn2019Below25Yards.groupby(['season','possessionTeam','gameId','playId','quarter','down','yardsToGo','yardlineSide','Result','gameClock',
                                   'yardlineNumber','kickReturnYardage','playResult','preSnapHomeScore','preSnapVisitorScore','event','homeTeamAbbr','visitorTeamAbbr',
                                   'week'],as_index=True).size()
LAKickoffReturn2019Below25YardsFrame = pd.DataFrame(LAKickoffReturn2019Below25YardsSeries)

mergedLAKickoffReturn2019Below25YardsSeries = mergedLAKickoffReturn2019Below25Yards.groupby(
    ['season','kickoffReturnFormation','tackler','gameId','playId','returnDirectionIntended',
     'returnDirectionActual','kickReturnYardage','homeTeamAbbr','visitorTeamAbbr','week'],as_index=True).size()
mergedLAKickoffReturn2019Below25YardsFrame = pd.DataFrame(mergedLAKickoffReturn2019Below25YardsSeries)

# Below 25 Yards 2020
LAKickoffReturn2020Below25YardsSeries = LAKickoffReturn2020Below25Yards.groupby(['season','possessionTeam','gameId','playId','quarter','down','yardsToGo','yardlineSide','Result','gameClock',
                                   'yardlineNumber','kickReturnYardage','playResult','preSnapHomeScore','preSnapVisitorScore','event','homeTeamAbbr','visitorTeamAbbr',
                                   'week'],as_index=True).size()
LAKickoffReturn2020Below25YardsFrame = pd.DataFrame(LAKickoffReturn2020Below25YardsSeries)

mergedLAKickoffReturn2020Below25YardsSeries = mergedLAKickoffReturn2020Below25Yards.groupby(
    ['season','kickoffReturnFormation','tackler','gameId','playId','returnDirectionIntended',
     'returnDirectionActual','kickReturnYardage','homeTeamAbbr','visitorTeamAbbr','week'],as_index=True).size()
mergedLAKickoffReturn2020Below25YardsFrame = pd.DataFrame(mergedLAKickoffReturn2020Below25YardsSeries)

# Loosing Yards 2020
LAKickoffReturn2019LossYardsSeries = LAKickoffReturn2019LossYards.groupby(['season','possessionTeam','gameId','playId','quarter','down','yardsToGo','yardlineSide','Result','gameClock',
                                   'yardlineNumber','kickReturnYardage','playResult','preSnapHomeScore','preSnapVisitorScore','event','homeTeamAbbr','visitorTeamAbbr',
                                   'week'],as_index=True).size()
LAKickoffReturn2019LossYardsFrame = pd.DataFrame(LAKickoffReturn2019LossYardsSeries)

mergedLAKickoffReturn2019LossYardsSeires = mergedLAKickoffReturn2019LossYards.groupby(
    ['season','kickoffReturnFormation','specialTeamsSafeties','gameId','playId','returnDirectionIntended',
     'returnDirectionActual','kickReturnYardage','homeTeamAbbr','visitorTeamAbbr','week'],as_index=True).size()
mergedLAKickoffReturn2019LossYardsFrame = pd.DataFrame(mergedLAKickoffReturn2019LossYardsSeires)
```
