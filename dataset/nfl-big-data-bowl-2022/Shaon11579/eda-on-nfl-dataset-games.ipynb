**Description of Game Data** : Primary Key is gameId

* **gameId**: Game identifier, unique (numeric)

* **gameDate**: Game Date (time, mm/dd/yyyy)

* **gameTimeEastern**: Start time of game (time, HH:MM:SS, EST)

* **homeTeamAbbr**: Home team three-letter code (text)

* **visitorTeamAbbr**: Visiting team three-letter code (text)

* **week**: Week of game (numeric)

```python
games.shape
```

```python
def Familiar_with_feature(data):
    df=pd.DataFrame(data.dtypes, columns=['Data Type'])
    df = df.reset_index()
    df= df.rename(columns={'index ': 'features'})
    df['Total Null'] = data.isnull().sum().values
    df['Unique Values'] =data.nunique().values
    df['First Value'] = data.iloc[0].values
    df['Lasrt Values'] = data.iloc[len(data)-1].values
    return df

Familiar_with_feature(games)
```

It consists of 272 games, with each of the NFL's 32 teams playing 17 games during an 18-week period with one "bye" week off. Since 2012, the NFL generally schedules games in five time slots during the week

```python
colors = ['gold','lightskyblue', 'green']
games['season'].value_counts().plot.pie(figsize=(10,14),colors=colors,shadow=True,explode=(0.2,0.1,0.2),autopct="%1.1f%%")
plt.title('Number of game playing for each season',fontsize=15)
plt.show()
```

```python
import plotly.express as px
check = games['gameDate'].value_counts().reset_index()

check.columns = [
    'date', 
    'games'
]

check = check.sort_values('games')

fig = px.bar(
    check, 
    y='date', 
    x="games", 
    orientation='h', 
    color='date',
    title='Number of games for every date', 
    height=900, 
    width=800
)
fig.show()
```

```python
import plotly.express as px
check = games['week'].value_counts().reset_index()

check.columns = [
    'date', 
    'games'
]

check = check.sort_values('games')

fig = px.bar(
    check, 
    y='date', 
    x="games", 
    orientation='h', 
    color='date',
    text='games',
    title='Number of games for per week', 
    height=900, 
    width=800
)
fig.show()
```

```python
import plotly.express as px
check = games['Home_team_sign'].value_counts().reset_index()

check.columns = [
    'date', 
    'games'
]

check = check.sort_values('games')

fig = px.bar(
    check, 
    y='date', 
    x="games", 
    orientation='h', 
    color='date',
    title='Number of home games', 
    text='games',
    height=900, 
    width=800
)
fig.show()
```
