```python
track_test = pd.read_csv("/kaggle/input/nfl-big-data-bowl-2022/tracking2018.csv")
```

```python
track_test.head()
```

```python
play_test = track_test.loc[(track_test['gameId'] == 2018123000) & (track_test['playId'] == 36)]
```

```python
play_test.shape
```

```python
import matplotlib as mlp
import matplotlib.pyplot as plt
```

```python
play_frame2 = play_test.loc[(play_test['time'] == '2018-12-30T21:25:38.900')]
play_frame2['team'] = (play_frame2['team'] != 'home').astype(int)
x = play_frame2['x']
y = play_frame2['y']
color = play_frame2['team']
```

```python
# general plot of field

fig, ax = plt.subplots(figsize=(12, 6))

plt.xlim([0, 120])
plt.ylim([0, 53.3])

plt.axvline(x=10, c='r')
plt.axvline(x=20, c='k')
plt.axvline(x=30, c='k')
plt.axvline(x=40, c='k')
plt.axvline(x=50, c='k')
plt.axvline(x=60, c='g')
plt.axvline(x=70, c='k')
plt.axvline(x=80, c='k')
plt.axvline(x=90, c='k')
plt.axvline(x=100, c='k')
plt.axvline(x=110, c='r')

plt.scatter(x,y, c=color)
```

```python
import plotly.express as px

px.scatter(play_test, x='x', y='y', animation_frame='time', color='team', range_x=[0, 120], range_y=[0, 53.3])
```

```python
def play_plotter(season_df, gameID, playID):
    
    play = season_df.loc[(season_df['gameId'] == gameID) & (track_test['playId'] == playID)]
    
    return px.scatter(play, x='x', y='y', animation_frame='time', color='team', range_x=[0, 120], range_y=[0, 53.3])
```

```python
play_plotter(track_test, 2018123000, 36)
```
