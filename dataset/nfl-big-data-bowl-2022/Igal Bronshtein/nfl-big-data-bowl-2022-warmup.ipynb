#### Height appear in two different ways. Converting them all into centimeters.

```python
height_dict = {
    "5-10": 178,
    "5-11": 180,
    "5-6": 168,
    "5-7": 170,
    "5-8": 173,
    "5-9": 175,
    "6-0": 183,
    "6-1": 185,
    "6-2": 188,
    "6-3": 190,
    "6-4": 193,
    "6-5": 196,
    "6-6": 198,
    "6-7": 201,
    "6-8": 203,
    "6-9": 206,
    "66": 168,
    "67": 170,
    "68": 173,
    "69": 175,
    "70": 178,
    "71": 180,
    "72": 183,
    "73": 185,
    "74": 188,
    "75": 191,
    "76": 193,
    "77": 196,
    "78": 198,
    "79": 201
}
```

```python
df["height_cm"] = df["height"].replace(height_dict)
df["weight_kg"] = df["weight"] * 0.4535924
df["weight_kg"] = df["weight_kg"].astype(int)
```

```python
# Body mass index
df["bmi"] = df["weight_kg"] / ((df["height_cm"]/100)* (df["height_cm"]/100))
df["bmi"] = df["bmi"].round(1)
```

```python
df['DateTime'] = pd.to_datetime(df['birthDate'],errors='coerce')
df["dayofweek"] = df["DateTime"].dt.day_name()
df["month"] =  df["DateTime"].dt.month
df["year"] = df["DateTime"].dt.year
CURRENT_YEAR = 2021
df["age"] = CURRENT_YEAR - df["year"]
```
