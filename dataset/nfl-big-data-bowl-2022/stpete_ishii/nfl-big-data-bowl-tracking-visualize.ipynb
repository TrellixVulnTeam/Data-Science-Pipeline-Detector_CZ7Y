```python
tracking2020=pd.read_csv('../input/nfl-big-data-bowl-2022/tracking2020.csv')
tracking2020[0:3]
```

```python
track1=tracking2020[tracking2020['playId']==40]
track1[0:3]
```

```python
fig=make_subplots(specs=[[{"secondary_y":False}]])
fig.add_trace(go.Scatter(x=track1['x'],y=track1['y'],name='track1'),secondary_y=False,)
fig.update_layout(autosize=False,width=700,height=500,title_text="track1")
fig.update_xaxes(title_text="x")
fig.update_yaxes(title_text="y")
fig.show()
```

```python
Name0=track1['jerseyNumber'].unique().astype(int).tolist()
Name=sorted(Name0)
print(Name)
```

```python
fig=make_subplots(specs=[[{"secondary_y":False}]])
for i in range(1,11):
    fig.add_trace(go.Scatter(x=track1[track1['jerseyNumber']==i]['x'],y=track1[track1['jerseyNumber']==i]['y'],name='jerseyNumber: '+str(i)),secondary_y=False,)
fig.update_layout(autosize=False,width=700,height=500,title_text="track1")
fig.update_xaxes(title_text="x")
fig.update_yaxes(title_text="y")
fig.show()
```

```python
Time0=track1['time'].unique().tolist()
Time=sorted(Time0)
print(Time[0:10])
```

```python
fig=make_subplots(specs=[[{"secondary_y":False}]])
for i in range(1,30,2):
    fig.add_trace(go.Scatter(x=track1[track1['time']==Time[i]]['x'],y=track1[track1['time']==Time[i]]['y'],mode='markers',name='time: '+str(Time[i])),secondary_y=False,)
    fig.update_layout(autosize=False,width=700,height=500,title_text="track1")
    fig.update_xaxes(title_text="x")
    fig.update_yaxes(title_text="y")
fig.show()
```
