```python
import matplotlib.pyplot as pyplot
import seaborn as sns
from sklearn.feature_extraction.text import CountVectorizer
from wordcloud import WordCloud
from collections import Counter
import matplotlib.pyplot as plt
```

```python
train_df = pd.read_csv('/kaggle/input/commonlitreadabilityprize/train.csv')
train_df.head()
```

```python
train_df['excerpt'][0]
```

```python
train_df.shape, train_df.describe()
```

```python
train_df['excerpt_split'] = train_df.excerpt.str.split(' ')
```

```python
all_words = []
[all_words.extend(a) for a in train_df['excerpt_split'].values]

pass
```

```python
sns.histplot(train_df.excerpt_split.apply(lambda x:len(x)).values)
plt.xlabel("length of sentences")
plt.ylabel("Count of occurances")
plt.title("sentences length vs their occurances")
plt.show()
```

```python
c = Counter(all_words)
```

```python
c.most_common(10)
```

```python
wc = WordCloud(max_font_size=50, max_words=100,
              background_color='black').generate_from_frequencies(c)

plt.imshow(wc, interpolation='bilinear')

plt.axis('off')
plt.show()
```
