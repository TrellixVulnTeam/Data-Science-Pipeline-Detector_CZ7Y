## Importing Libraries

```python
from fastai.text.all import *
import plotly.express as px
import plotly.figure_factory as ff
from wordcloud import WordCloud, STOPWORDS
```

## Data Exploration

```python
data = Path('../input/commonlitreadabilityprize')
train = pd.read_csv(data/'train.csv')
# test = pd.read_csv(data/'test.csv')
# sample = pd.read_csv(data/'sample_submission.csv')
```

```python
train.info()
```

- Lots of null Values in url_legal and license columns

```python
length = [len(x.split(' ')) for x in train['excerpt']]
px.density_heatmap(y=train['target'], x=length, labels=dict(x = 'Length of Text (words)', y = 'Reading Ease (Target)'))
```

```python
train.corr()
```

- Almost no correalation between length of words and Ease of Readability is observed

```python
px.scatter(y=train['target'], x=train['standard_error'], marginal_x='histogram', marginal_y='histogram', opacity=0.2, labels = dict(x = 'Standard Error', y = 'Reading Ease (Target)'))
```

- ### Interesting results... this indicates the scores are the most spread out when the text is either very easy to read or very hard to read

```python
train.describe()
```

# Visualising The Texts With Lowest Ease of Readability

```python
df = train.iloc[:500]
standard_errors = []
```

## Top 500 Hardest To Read Texts

```python
stopwords = set(STOPWORDS)
wc = WordCloud(background_color="white", max_words=200, stopwords=stopwords, width = 1500, height = 750)

plt.figure(figsize=(16,16))
plt.imshow(wc.generate(' '.join(df['excerpt'])))
```

```python
standard_errors.append(list(df['standard_error']))
```

## Top 500 - 1000 Hardest To Read Texts

```python
df = train.iloc[500:1000]

plt.figure(figsize=(16,16))
plt.imshow(wc.generate(' '.join(df['excerpt'])))
```

```python
standard_errors.append(list(df['standard_error']))
```

# Visualising The Texts With Highest Ease of Readability

# Easiest 500 To Read

```python
df = train.iloc[-500:]

plt.figure(figsize=(16,16))
plt.imshow(wc.generate(' '.join(df['excerpt'])))
```

```python
standard_errors.append(list(df['standard_error']))
```

## Easiest 500 - 1000 to Read

```python
df = train.iloc[-1000:-500]

plt.figure(figsize=(16,16))
plt.imshow(wc.generate(' '.join(df['excerpt'])))
```

```python
standard_errors.append(list(df['standard_error']))
```

```python
group_labels = ['Top 500 Toughest to Read', 'Top 500 - 1000 Toughest to Read', 'Top 500 Easiest to Read', 'Top 500 - 1000 Easiest To Read']
fig = ff.create_distplot(standard_errors, group_labels, bin_size=0.007)
fig.update_xaxes(range=[0.43, 0.65])
```
