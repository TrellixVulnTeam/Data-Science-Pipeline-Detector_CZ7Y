```python
import spacy
```

```python
train = pd.read_csv('/kaggle/input/commonlitreadabilityprize/train.csv')
train
```

```python
test=pd.read_csv('/kaggle/input/commonlitreadabilityprize/test.csv')
test
```

## Vectorization of text for training

```python
nlp = spacy.load('en_core_web_lg')
with nlp.disable_pipes():
    train_vectors = np.array([nlp(text).vector for text in train.excerpt])
    
train_vectors.shape
```

## Train test split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(train_vectors, train.target,
                                                    test_size=0.1, random_state=1)
```

## Model

```python
from sklearn.linear_model import LinearRegression

model=LinearRegression()
model.fit(X_train, y_train)
```

## Vectorization of testset for testing

```python
with nlp.disable_pipes():
    test_vectors = np.array([nlp(text).vector for text in test.excerpt])
    
test_vectors.shape
```

## Predicting the target

```python
test['target'] = model.predict(test_vectors)
```
