Synopsis
===
Barebones autokeras regression.

Heavily influenced by https://www.kaggle.com/heyytanay/auto-keras-text-regression-starter/data


Install autokeras
===

```python
import sys
package_dir = '../input/autokeras-april-2021'
sys.path.insert(0, package_dir)
```

Load the data into pandas
===

```python
import pandas as pd

data_train = pd.read_csv("/kaggle/input/commonlitreadabilityprize/train.csv")
data_test = pd.read_csv("/kaggle/input/commonlitreadabilityprize/test.csv")

data_train.head()
```

Add beginning of string and end of string tokens
===

```python
data_train["excerpt"] = data_train["excerpt"].apply(lambda x: "<bos> "+x+" <eos>")
data_test["excerpt"] = data_test["excerpt"].apply(lambda x: "<bos> "+x+" <eos>")
```

Extract features
===

```python
text_train = data_train["excerpt"].values
text_test = data_test["excerpt"].values
text_train.shape
```


Extract labels
===

```python
labels_train = data_train["target"].to_numpy()
```

Building and training the model
===

```python
import autokeras

text_regressor = autokeras.TextRegressor(
    #loss="mean_squared_error",
    max_trials=25,
    #objective="val_loss",
    overwrite=True
)

text_regressor.fit(x=text_train, y=labels_train, validation_split=0.1)
```

Predictions
===

```python
import numpy as np


test_predictions = text_regressor.predict(text_test).flatten()
np.set_printoptions(suppress=True)
print(test_predictions)
```

Submission
===

```python
data_to_submit = pd.DataFrame({
    'id':data_test['id'],
    'target': test_predictions
})
data_to_submit.to_csv('submission.csv', index = False)
data_to_submit
```
