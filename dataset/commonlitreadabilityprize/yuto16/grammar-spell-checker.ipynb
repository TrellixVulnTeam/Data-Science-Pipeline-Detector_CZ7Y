# Check all train excerpts and calculate correlation with target

```python
train["GS_check"] = [tool.check(text) for text in tqdm(train["excerpt"])]
```

```python
train["GS_check_category"] = [[m.category for m in matches] for matches in train["GS_check"]]

train["GS_check_cnt_GRAMMAR"] = [cate.count("GRAMMAR") for cate in train["GS_check_category"]]
train["GS_check_cnt_TYPOS"] = [cate.count("TYPOS") for cate in train["GS_check_category"]]
train["GS_check_cnt_TYPOGRAPHY"] = [cate.count("TYPOGRAPHY") for cate in train["GS_check_category"]]
train["GS_check_cnt_REDUNDANCY"] = [cate.count("REDUNDANCY") for cate in train["GS_check_category"]]
train["GS_check_cnt_PUNCTUATION"] = [cate.count("PUNCTUATION") for cate in train["GS_check_category"]]
train["GS_check_cnt_STYLE"] = [cate.count("STYLE") for cate in train["GS_check_category"]]
train["GS_check_cnt_MISC"] = [cate.count("MISC") for cate in train["GS_check_category"]]
train["GS_check_cnt_CASING"] = [cate.count("CASING") for cate in train["GS_check_category"]]

colnames = ["GS_check_cnt_GRAMMAR", "GS_check_cnt_TYPOS", "GS_check_cnt_TYPOGRAPHY", "GS_check_cnt_REDUNDANCY", 
            "GS_check_cnt_PUNCTUATION", "GS_check_cnt_STYLE", "GS_check_cnt_MISC", "GS_check_cnt_CASING", "target"]
```

```python
train.loc[:,colnames].head()
```

```python
corr = train[colnames].corr()

fig = plt.figure(figsize=(12,12),dpi=80)
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, cmap='BuPu', robust=True, center=0,
            square=True, linewidths=.5, annot=True)
plt.show()
```
