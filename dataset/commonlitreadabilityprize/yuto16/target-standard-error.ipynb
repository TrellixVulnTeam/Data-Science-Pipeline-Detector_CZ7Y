```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv("/kaggle/input/commonlitreadabilityprize/train.csv")
```

# scatter plot (target & standard_error)

```python
plt.scatter(train["target"], train["standard_error"])
plt.xlabel("target")
plt.ylabel("standard_error")
plt.show()
```

```python
plt.scatter(train["target"], train["standard_error"])
plt.xlabel("target")
plt.ylabel("standard_error")
plt.plot([-1,-1],[0.4,0.7], color="gray", linestyle="dashed")
plt.ylim([0.4,0.7])
plt.show()
```
