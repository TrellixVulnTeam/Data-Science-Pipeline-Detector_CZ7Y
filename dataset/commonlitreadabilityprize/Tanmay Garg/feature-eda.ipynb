```python
# just importing required libraries here.
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
```

```python
#FYI, df means pandas dataframe.
train_df = pd.read_csv('../input/commonlitreadabilityprize/train.csv')
train_df.info()
```

# 1. Relation between excerpt word count and target

Let's see if:
1. reading ease (target)
2. standard error
values vary with the **excerpt word count** by fitting a simple linear regression model.

```python
#FYI, dfs means pandas series.
excerpt_word_count_dfs = train_df.excerpt.str.split().apply(len)
```

```python
ax = sns.regplot(x=excerpt_word_count_dfs,y=train_df.target,marker='.',scatter_kws={"s": 2})
ax.set(xlabel ='excerpt word count')
```

Higher word counts are annotated with lower complexities.

```python
sns.regplot(x=excerpt_word_count_dfs,y=train_df.standard_error,marker='.',scatter_kws={"s": 2})
ax.set(xlabel ='excerpt word count')
```

The line is almost horizontal here, meaning the inter-annotator agreement does not vary with length.
