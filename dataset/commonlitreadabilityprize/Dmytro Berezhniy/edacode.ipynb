# **Looking at excerpts**

# **Excerpts length (symbols)** 

```python
fig = px.histogram(data_train,x = data_train['excerpt'].apply(len),
                  title = 'Except length(symbols) frequency',
                  color_discrete_sequence = ['royalblue'])
fig.update_layout(
xaxis_title_text = 'Except length(symbols)',
yaxis_title_text = 'Frequency',
bargap = 0.25, showlegend = False,
             autosize=False,
             width=750,
             height=450)
```

# **Unique symbols**

```python
all_chars = set()
data_train['excerpt'].apply(lambda x: [all_chars.add(c) for c in x])
for c in sorted(all_chars):
    print(c + ' ', end='')
    
print('\n\n')
    
for c in sorted(all_chars):
    print(f'({c}, {str(ord(c))}) ', end='')
```

# **Creating column with boolean value of existing unusual symbols**

```python
hard_chars = set()
for c in all_chars:
    if ord(c) >= 176 and ord(c) <= 339:
        hard_chars.add(c)
        
print(hard_chars)

data_train['unusual_symb'] = data_train['excerpt'].apply(lambda x: any([c in hard_chars for c in x]))
print("Number of examples with unusual symbols:",sum(data_train['unusual_symb']))
```
