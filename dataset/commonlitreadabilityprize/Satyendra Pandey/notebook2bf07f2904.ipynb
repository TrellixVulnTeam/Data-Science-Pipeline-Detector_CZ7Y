```python
dataset = pd.read_csv('/kaggle/input/commonlitreadabilityprize/train.csv')
```

```python
dataset
```

```python
X = dataset.copy()
X.drop(['url_legal','license','id','target'], axis = 'columns', inplace = True)
```

```python
y = dataset['target'].copy()
```

```python
data = X['excerpt']
data.replace(['^a-z^A-Z'],' ', regex = True, inplace = True)
```

```python
data
```

```python

import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
```

```python
for i in range(0, len(X)):
  review = re.sub('[^a-zA-Z]', ' ', X['excerpt'][i])
  review = review.lower()
  review = review.split()
  review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
  review = ' '.join(review)
  corpus.append(review)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 409)
data_n = cv.fit_transform(corpus).toarray()
```

```python
data_n
```

```python
y_n = y.to_numpy()
```

```python
y_n
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data_n, y_n, test_size = 0.80, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```
