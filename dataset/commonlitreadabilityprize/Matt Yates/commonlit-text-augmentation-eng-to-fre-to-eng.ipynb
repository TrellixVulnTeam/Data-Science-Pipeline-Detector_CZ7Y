# English --> French --> English

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
from textblob import TextBlob
import time
```

```python
df = pd.read_csv("../input/commonlitreadabilityprize/train.csv")
df.head()
```

Need to space out the Google Translate API calls otherwise "too many requests" error is returned

```python
df['augtext'] = [''] * len(df)

for index, row in df.iterrows():
    french_translation = str(TextBlob(row['excerpt']).translate(to='fr'))
    df.loc[index,'augtext'] = str(TextBlob(french_translation).translate(to='en')) # back to English
    time.sleep(10)
    
df.head()
```

```python
print("--- Example Translation ---")
print("")
print("Original Text: ", df.values[0,3])
print("")
print("Augmented Text: ", df.values[0,6])
print("")
```

```python
df.to_csv("commonlit_train_with_augs.csv", index=False)
```
