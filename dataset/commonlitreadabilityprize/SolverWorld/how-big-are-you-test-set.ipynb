Read in the train and test sets, count how many entries are in each

```python
train = pd.read_csv('/kaggle/input/commonlitreadabilityprize/train.csv')
test  = pd.read_csv('/kaggle/input/commonlitreadabilityprize/test.csv')

num_train=len(train)
num_test=len(test)
print(f'number of train examples={num_train}, number of test examples={num_test}')
```

Now prepare a test submission with every value set to the number of test examples.  Given that target values are approximately -4 to 1, the RMSE of this entry should be approximately equal to the number of test examples (assuming there are more than 100 or so).

```python
test['final_prediction']=num_test   # make a column of with all values the same = num_test
submit_df = pd.DataFrame({'id':test.id,'target':test['final_prediction']})
print(submit_df)
submit_df.to_csv("submission.csv", index = False)
```
