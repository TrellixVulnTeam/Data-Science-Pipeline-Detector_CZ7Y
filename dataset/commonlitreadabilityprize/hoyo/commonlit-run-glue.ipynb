```python
import pandas as pd
import numpy as np

# The script needs only the text data so we have to make new data frames with only the 
# excerpts

train = pd.read_csv('../input/commonlitreadabilityprize/train.csv')
train_texts = train[['excerpt','target']].copy()
train_texts.rename(columns={'excerpt':'text', 'target':'label'}, inplace=True)
train_texts.to_csv('train_data.csv', index=False)
```
