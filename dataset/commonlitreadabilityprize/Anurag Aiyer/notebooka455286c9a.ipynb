```python
###Create functions for readability metrics
## Make sure to have all the necessary components
import re
import requests
import math
import nltk
from collections import Counter
from nltk import word_tokenize, pos_tag

def count_syllables(excerpt):
    word_array = re.split('\s+', excerpt)
    total_syllables = 0
    for word in word_array:
        total_syllables = total_syllables + len(
        re.findall('(?!e$)[aeiouy]+', word, re.I) +
        re.findall('^[^aeiouy]*e$', word, re.I))
    return total_syllables

def poly_syllables(excerpt):
    word_array = re.split('\s+', excerpt)
    count = 0
    for word in word_array:
        total_syllables = len(
        re.findall('(?!e$)[aeiouy]+', word, re.I) +
        re.findall('^[^aeiouy]*e$', word, re.I))
        if total_syllables > 1:
            count += 1
        
    
    return count
            
    return total_syllables
def total_words(excerpt):
    res = len(excerpt.split())
    return res

def count_sentences(excerpt):
    sentence_count = 0
    seen_end = False
    sentence_end = {'?', '!', '.'}
    for c in excerpt:
        if c in sentence_end:
            if not seen_end:
                seen_end = True
                sentence_count += 1
            continue
        seen_end = False
    return sentence_count
        
def average_words_per_sentence(excerpt):
    average_count = total_words(excerpt) / count_sentences(excerpt)
    return average_count

def long_words(excerpt):
    output = []
    
    for x in excerpt.split():
        if len(x) >= 10:
            output.append(x)
    long_words = len(output)   
    return long_words

def percent_long_words(excerpt):
    percent_long = long_words(excerpt) / total_words(excerpt)
    return percent_long

def difficult_words(excerpt):
    

    english_most_common_10k = 'https://raw.githubusercontent.com/first20hours/google-10000-english/master/google-10000-english-usa-no-swears.txt'

    # Get the file of 10 k most common words from TXT file in a github repo
    response = requests.get(english_most_common_10k)
    data = response.text
    difficult_words = []

    set_of_common_words = {x for x in data.split('\n')}

    # Once we have the set of common words, we can just check.
    # The check is in average case O(1) operation,
    # but you can use for example some sort of search three with O(log(n)) complexity
    for word in excerpt.split(): 
        if word not in set_of_common_words:
            difficult_words.append(word)
    difficult_len = len(difficult_words)
    return difficult_len

            
def average_sentence_in_words(excerpt):
    sents = excerpt.split('.')
    avg_len = sum(len(x.split()) for x in sents) / len(sents)
    return avg_len

def count_chars(excerpt):
    result = 0
    excerpt = excerpt.replace('.','')
    for char in excerpt:
        result += 1     # same as result = result + 1
    return result

def modal_verbs(excerpt):
    modals = ['can', 'could', 'may', 'might', 'must', 'will', 'would', 'should']
    excerpt = excerpt.replace('.','')
    count = 0
    for x in excerpt:
        if x.lower() in modals:
            count=count + 1
    return count
            
### determiners, noun, and verbs
def dt_noun_verb_counts(excerpt):
    dt = [word for word, pos in pos_tag(word_tokenize(excerpt)) if pos.startswith('DT')]
    vb = [word for word, pos in pos_tag(word_tokenize(excerpt)) if pos.startswith('VB')]
    nn = [word for word, pos in pos_tag(word_tokenize(excerpt)) if pos.startswith('NN')]
    return len(dt), len(vb), len(nn)

def percent_difficult_words(excerpt):
    text_input_words = excerpt.split()
    difficult_words_count = 0
    
    for i in text_input_words:
        if i not in dale_chill_list:
            difficult_words_count += 1
    return difficult_words_count/ len(text_input_words)
```

## We will create Fleasch Reading Ease, LIX, DaleChillIndex, Kincaid, Coleman-Liau, Rix, and Gunning Fox Index functions to use as part of our features for the processed training model

```python
def fleach_ease(excerpt):
    score = 206.835 - (1.015*(total_words(excerpt)/count_sentences(excerpt))) - (84.6 *(count_syllables(excerpt)/total_words(excerpt)))
    return score

def smog_index(excerpt):
    score = 3 + math.sqrt(poly_syllables(excerpt))
    return score
def lix(excerpt):
    score = percent_long_words(excerpt) + average_words_per_sentence(excerpt)
    return score 
def kincaid(excerpt):
    score = .39 * (total_words(excerpt)/count_sentences(excerpt)) + (11.8 * (count_syllables(excerpt)/total_words(excerpt))) - 15.59
    return score

def ari(excerpt):
    score = 4.71 * (count_chars(excerpt)/total_words(excerpt)) + (0.5 * (total_words(excerpt)/count_sentences(excerpt))) - 21.43
    return score
def coleman_liau(excerpt):   
    l = 100 * count_chars(excerpt) / total_words(excerpt);
    s = 100 * count_sentences(excerpt) / total_words(excerpt);
    score = 0.0588 * l - 0.296 * s - 15.8
    return score

def rix(excerpt):   
    score = long_words(excerpt) / count_sentences(excerpt)
    return score
def gunning_fog_index(excerpt):
    score = ((total_words(excerpt)/count_sentences(excerpt)) + 100*(count_syllables(excerpt)/total_words(excerpt)))*0.4
    return score

def dale_chall(excerpt):
    score = 0.1579 * (percent_difficult_words(excerpt)) + (0.0496 * average_sentence_in_words(excerpt))
    if percent_difficult_words(excerpt) > .05:
        final_score = score + 3.6365
        return final_score
    else:
        return final_score
    
```

```python
import string
def clean_text(text):
    '''Make text lowercase, remove text in square brackets,remove links,remove punctuation
    and remove words containing numbers.'''
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text


def text_preprocessing(text):
    """
    Cleaning and parsing the text.

    """
    tokenizer = nltk.tokenize.RegexpTokenizer(r'\w+')
    nopunc = clean_text(text)
    tokenized_text = tokenizer.tokenize(nopunc)
    #remove_stopwords = [w for w in tokenized_text if w not in stopwords.words('english')]
    combined_text = ' '.join(tokenized_text)
    return combined_text
```
