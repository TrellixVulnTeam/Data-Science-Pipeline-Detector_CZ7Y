# About this notebook

* AWD-LSTM model finetuned on the competition dataset.
* Inference notebook for the Fastai training [notebook](https://www.kaggle.com/snnclsr/commonlit-fastai-training-with-language-model)

Simply loads the trained model and make predictions on the test dataset.

```python
from pathlib import Path

from fastai.text.all import *
```

```python
is_cpu = False if torch.cuda.is_available() else True

df_test = pd.read_csv("../input/commonlitreadabilityprize/test.csv")

learn = load_learner("../input/common-lit-fastai-awd-lstm/final.pkl", cpu=is_cpu)
```

```python
test_dl = learn.dls.test_dl(df_test.excerpt)
preds, _ = learn.get_preds(dl=test_dl)
```

```python
df_sub = pd.read_csv("../input/commonlitreadabilityprize/sample_submission.csv")
df_sub['target'] = preds.numpy()
df_sub.to_csv('submission.csv',index=False)
```
