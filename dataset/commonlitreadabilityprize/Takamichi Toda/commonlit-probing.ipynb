```python
import numpy as np
from sklearn.metrics import mean_squared_error
```

```python
train = pd.read_csv(f"/kaggle/input/commonlitreadabilityprize/train.csv")
```

```python
all_0 = mean_squared_error(train['target'], np.ones(len(train))-1, squared=False)
all_1 = mean_squared_error(train['target'], np.ones(len(train)), squared=False)
all_2 = mean_squared_error(train['target'], np.ones(len(train))+1, squared=False)
all_3 = mean_squared_error(train['target'], np.ones(len(train))+2, squared=False)
all_minus1 = mean_squared_error(train['target'], np.ones(len(train))-2, squared=False)
all_minus2 = mean_squared_error(train['target'], np.ones(len(train))-3, squared=False)
all_minus3 = mean_squared_error(train['target'], np.ones(len(train))-4, squared=False)
```

```python
print(f'all  3 = {all_3:0.4} / 4.083')
print(f'all  2 = {all_2:0.4} / 3.127')
print(f'all  1 = {all_1:0.4} / 2.210')
print(f'all  0 = {all_0:0.4} / 1.412')
print(f'all -1 = {all_minus1:0.4} / 1.049')
print(f'all -2 = {all_minus2:0.4} / 1.486')
print(f'all -3 = {all_minus3:0.4} / 2.305')
```
