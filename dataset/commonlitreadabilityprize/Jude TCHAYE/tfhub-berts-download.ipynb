# Download encoders

```python
! mkdir /kaggle/tmp/data/encoder/
```

```python
for key in map_name_to_handle:
    path = map_name_to_handle[key]
    idx = key.rfind('/')
    if idx > 0:
        os.makedirs(f"/kaggle/tmp/data/encoder/{key[:idx]}",exist_ok=True)        
    print(key,path)
    wget.download(f'{path}?tf-hub-format=compressed', f'/kaggle/tmp/data/encoder/{key}.tar.gz')
    tar = tarfile.open(f'/kaggle/tmp/data/encoder/{key}.tar.gz', "r:gz")
    tar.extractall(path=f'/kaggle/tmp/data/encoder/{key}')
    tar.close()
    
    os.remove(f'/kaggle/tmp/data/encoder/{key}.tar.gz')
```

# Download preprocessors

```python
! mkdir /kaggle/tmp/data/preprocess/
```
