```python
#Importing the mandatory libraries
import pandas as pd 
import numpy as np 
import tensorflow as tf 
import nltk
from nltk.stem import PorterStemmer
```

```python
#Reading the data
train_df=pd.read_csv('../input/nlp-getting-started/train.csv', header='infer')
test_df=pd.read_csv('../input/nlp-getting-started/test.csv', header='infer')
```

```python
# Extracting only the needed column
train_df = train_df[['text']]
test_df = test_df[['text']]
```

```python
train_df.head()
```

```python
# Converting into lower case
train_df['text'] = train_df['text'].str.lower()
train_df.head()
```

```python
# Removing the special symbols
import re
for i in range(len(train_df)):
    train_df.iloc[i][0] = re.sub('[^A-Za-z0-9]+', ' ', train_df.iloc[i][0])
```

```python
train_df.head
```

```python
# Tokenizing and forming the vocabulary
for i in range(len(train_df)):
    train_df.iloc[i][0] = nltk.word_tokenize(train_df.iloc[i][0])
```

```python
train_df.head
```

```python
# Stemming the vocabulary
ps = PorterStemmer()
train_df['text'] = train_df['text'].apply(lambda x: [ps.stem(y) for y in x]) # Stem every word.
```

```python
train_df.head
```

```python
Row_list =[]
  
# Iterate over each row
for index, rows in train_df.iterrows():
    my_list =[rows.text]
      
    Row_list.append(my_list)
```

```python
import itertools

merged = list(itertools.chain(*Row_list))
```

```python
words = list(itertools.chain.from_iterable(merged))
```

```python
word_str = ' '.join([str(elem) for elem in words])
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Generate a word cloud image
wordcloud = WordCloud().generate(word_str)

plt.imshow(wordcloud, interpolation="bilinear")
plt.show()
```

```python
from collections import Counter
import matplotlib.pyplot as plt


# Assign the Counter instance `most_common` call to a variable:
word_frequency = Counter(" ".join(words).split()).most_common(10)

# `most_common` returns a list of (word, count) tuples
all_words = [word for word, _ in word_frequency]
counts = [counts for _, counts in word_frequency]

plt.bar(all_words, counts)
plt.title("10 most frequent tokens in description")
plt.ylabel("Frequency")
plt.xlabel("Words")
plt.show()
```
