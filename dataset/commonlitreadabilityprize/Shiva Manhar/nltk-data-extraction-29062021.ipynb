# Flesch-Kincaid reading formula
![](https://i.ibb.co/xsYLbCk/formula1.png)

```python

def flesch_reading_level(total_words, total_sentence, total_syllables):
    flesch_level = 0.39*(total_words/total_sentence)+11.8*(total_syllables/total_words)-15.59
    return flesch_level
```

# Flesch-Dayani Score



![](https://i.ibb.co/s6FjDF5/formula2.png)

```python

def flesch_dayani_score(total_words, total_sentence, total_syllables):
    flesch_dayani = 0.31-8.846*(total_syllables/total_words)- 1.01*(total_words/total_sentence)
    return flesch_dayani
```

```python

def count_verb(wordlist):
    total_word = []
    verbs = []
    singular_nouns  =[]
    proper_nouns =[]
    adverb =[]
    tag = nltk.pos_tag(wordlist)
    grammar = "NP: {<RB.?>*<VB.?>*<NNP>*<NN>*}"
    cp  =nltk.RegexpParser(grammar)
    
    for w in list(tag):
        if w[1] == 'VB':
            verbs.append(len(w[0]))
        total_word.append(len(w[0]))
    return sum(verbs)/(sum(total_word)*100)
    #return len(verbs)*len(total_word)
 
    result = cp.parse(tag)
   
    for subtree in result.subtrees(filter=lambda t: t.label() == 'NP'):
        if 'VB' in subtree[0][1]:
            verbs.append(subtree[0][0])
            print(subtree[0][0])
        if 'RB' in subtree[0][1]:
            adverb.append(subtree[0][0])
        if 'NNP' in subtree[0][1]:
            proper_nouns.append(subtree[0][0])
        if 'NN' in subtree[0][1]:
            singular_nouns.append(subtree[0][0])
        return len(verbs)


df['verbs'] = list(map(count_verb, df.loc[:,'only_words']))
test['verbs'] = list(map(count_verb, test.loc[:,'only_words']))

new_df['verbs'] = list(map(count_verb, df.loc[:,'only_words']))
new_test['verbs'] = list(map(count_verb, test.loc[:,'only_words']))
```

```python

spell = SpellChecker()
def count_adverb(wordlist):
    total_word = []
    adverb = []
    tag = nltk.pos_tag(wordlist)
    #print(tag)
    for w in list(tag):
        if w[1] == 'RB':
            adverb.append(len(w[0]))
        total_word.append(len(w[0]))
    return sum(adverb)/(sum(total_word)*100)

def count_possessive_pronoun(wordlist):
    total_word = []
    pos_pronoun = []
    tag = nltk.pos_tag(wordlist)
    #print(tag)
    for w in list(tag):
        if w[1] == 'PRP$':
            pos_pronoun.append(len(w[0]))
        total_word.append(len(w[0]))
    return sum(pos_pronoun)

def count_wh_pronoun(wordlist):
    total_word = []
    wh_pronoun = []
    tag = nltk.pos_tag(wordlist)
    for w in list(tag):
        if w[1] == 'WP$':
            wh_pronoun.append(len(w[0]))
        total_word.append(len(w[0]))
    return sum(wh_pronoun)/(sum(total_word)*100)


def count_misspelled(wordlist, total_words):
    misspelled = spell.unknown(wordlist)
    return len(misspelled)/total_words*100
```

```python
def build_and_compile_model(norm):
    model = keras.Sequential([
        norm,   
        #layers.Dense(1024, activation='relu'),
        #layers.Dense(64, activation='relu'),
        #layers.Dense(64, activation='sigmoid'),
        #layers.Dense(128, activation='tanh'),
        #layers.Dense(1)
        layers.Dense(100, activation='relu'),
        layers.Dense(80, activation='relu'),
        layers.Dense(30, activation='relu'),
        layers.Dense(10, activation='relu'),
        layers.Dense(5, activation='relu'),
        layers.Dense(1)
    ])
    #opt =tf.keras.optimizers.SGD(learning_rate=0.01, momentum = 0.9)
    opt = tf.keras.optimizers.Adam(learning_rate=0.01)
    model.compile(loss='mse',   optimizer=opt)
    #model.compile(loss='sparse_categorical_crossentropy',   optimizer=tf.keras.optimizers.Adam(0.001))
    return model
```
