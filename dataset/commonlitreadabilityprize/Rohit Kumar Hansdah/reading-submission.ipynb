# Importing Libraries

```python
import pickle
import pandas as pd
from sklearn.ensemble import GradientBoostingRegressor
```

# Importing Data

```python
train_data=pd.read_csv(r'../input/commonlitreadabilityprize/train.csv')
train_data
```

```python
test_data=pd.read_csv(r'../input/commonlitreadabilityprize/test.csv')
test_data
```

```python
train_df=pd.read_csv(r'../input/data-for-model/train1.csv')
train_df=train_df.drop(columns=['target'],axis=1)
train_df
```

```python
test_df=pd.read_csv(r'../input/data-for-model/test1.csv')
test_df
```

# Model Training

```python
model = GradientBoostingRegressor(random_state=0)
```

```python
model.fit(train_df, train_data['target'])
```

# Predictions on Test Dataset

```python
predictions=model.predict(test_df)
predictions
```

# Submission

```python
ans=pd.DataFrame()
ans['id']=test_data['id']
ans['target']=predictions
ans.to_csv(r'./submission.csv',index=False)
```
