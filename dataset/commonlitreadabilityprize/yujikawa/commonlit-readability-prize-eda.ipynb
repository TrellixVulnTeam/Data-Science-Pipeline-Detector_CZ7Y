```python
# get stop word
nltk.download('stopwords')
stop_words = stopwords.words('english')
```

```python
def remove_symbol(text: str):
    text = text.replace(',', '')
    text = text.replace('.', '')
    text = text.replace('?', '')
    text = text.replace('!', '')
    return text.lower()
    
def excluded_stopwords(words):
    return [word.lower() for word in words if word.lower() not in stop_words]
```

```python
train['len_excerpt'] = train['excerpt'].apply(lambda x: len(remove_symbol(x).split()))
```

```python
train['excerpt_excluded_stopwords'] = train.excerpt.apply(lambda x: excluded_stopwords(remove_symbol(x).split()))
```

```python
train['len_excerpt_excluded_stopwords'] = train['excerpt_excluded_stopwords'].apply(lambda x: len(x))
```
