```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error as mse

file_path = "../input/train-and-test-csv/Data - final2.csv"
test_file_path = "../input/train-and-test-csv/Data - Sheet1.csv"
og_test_file_path = '../input/commonlitreadabilityprize/test.csv'

df = pd.read_csv(file_path)
test = pd.read_csv(test_file_path)
test_df = pd.read_csv(og_test_file_path)

y = df.target
features = ['longest_sentence', 'avgsentence', 'excerpt_length', 'avgword', 'unique_alphabets']
X = df[features]
z = test[features]

#train_X, test_X, train_y, test_y = train_test_split(X, y,random_state=1)

alpha_model = RandomForestRegressor()
alpha_model.fit(X, y)

alpha_predictions = alpha_model.predict(z)
predictions = pd.DataFrame()
predictions['id']=test_df['id']
predictions['target'] = alpha_predictions
predictions.to_csv('/kaggle/working/submission.csv', index=False)
predictions
```
