```python
import pandas as pd
from pycaret.regression import *

df1=pd.read_csv('../input/commonlitreadabilityprize/train.csv')
df2=pd.read_csv('../input/commonlitreadabilityprize/test.csv')
df3=pd.read_csv('../input/commonlitreadabilityprize/sample_submission.csv')

train=df1.drop(columns=['id', 'standard_error'])
test=df2.drop(columns=['id'])


exp_name = setup(data = train,  target = 'target')
```

```python
best=compare_models()
```

```python
best
```

```python
evaluate_model(best)
```

```python
finalyzed_best=finalize_model(best)
```
