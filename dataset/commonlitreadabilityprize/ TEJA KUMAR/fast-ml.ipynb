<h1 style="background-color:orange;font-family:newtimeroman;font-size:350%;text-align:center;border-radius: 15px 50px;">Missing Data Imputation for an numerical columns and its methods</h1>

```python
df1 = sns.load_dataset("titanic") 
df1.shape
```

```python
df1.head()
```

```python
df1.isnull().sum()
```

```python
ms = MissingDataImputer_Numerical(method="median")
```

```python
imputer = ms.fit(df1,["age"])
```

```python
df1 = imputer.transform(df1)
```

```python
df1.isnull().sum()
```

```python
cat_imputer = MissingDataImputer_Categorical(method = 'custom_value') 
cat_imputer.fit(df1,["embark_town"])  
df2 = cat_imputer.transform(df1) 
```

```python
ms1 = MissingDataImputer_Categorical(method='frequent') 
ms2 = ms1.fit(df1,["deck"]) 
ms1.transform(df1)
```

<h1 style="background-color:orange;font-family:newtimeroman;font-size:350%;text-align:center;border-radius: 15px 50px;">Outlier treatment by IQR</h1>

```python
plt.figure(figsize=(15,5))
df1.boxplot(column="age")
```

```python
ot = OutlierTreatment(method="gaussian")
```

```python
os = ot.fit(df1,variables=["age"])
```

```python
df2 = os.transform(df1)
```

```python
plt.figure(figsize=(15,5))
df2.boxplot(column="age")
```
