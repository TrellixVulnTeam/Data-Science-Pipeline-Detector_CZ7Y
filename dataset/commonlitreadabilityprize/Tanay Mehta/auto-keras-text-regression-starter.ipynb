```python
import os

import numpy as np
import pandas as pd
import tensorflow as tf

import autokeras as ak
```

```python
train_file = pd.read_csv("../input/commonlitreadabilityprize/train.csv")
train_file = train_file[['excerpt', 'target']]
train_file['excerpt'] = train_file['excerpt'].apply(lambda x: "<START> "+x+" <END>")

train = train_file[:2800]
valid = train_file[2800:]
```

```python
train_data = train['excerpt'].values
train_target = train['target'].values

valid_data = valid['excerpt'].values
valid_target = valid['target'].values

print(train_data.shape, train_target.shape)
print(valid_data.shape, valid_target.shape)
```

```python
# Initialize the text regressor.
reg = ak.TextRegressor(overwrite=True, max_trials=7)  # It tries 10 different models.
# Feed the text regressor with training data.
reg.fit(train_data, train_target, epochs=25)
# Predict with the best model.
predicted_y = reg.predict(valid_data)
# Evaluate the best model with testing data.
print(reg.evaluate(valid_data, valid_target))
```

```python
# Testing code
test_file = pd.read_csv("../input/commonlitreadabilityprize/test.csv")
test_ids = test_file['id'].tolist()
test_text = test_file['excerpt'].apply(lambda x: "<START> "+x+" <END>").values

test_file.head()
```

```python
# Get the predictions 
preds = reg.predict(test_text).squeeze().tolist()
```

```python
# Store into a dataframe
submission = pd.DataFrame()
submission['id'] = test_ids
submission['target'] = preds

# Save the dataframe
submission.to_csv("submission.csv", index=None)
```
