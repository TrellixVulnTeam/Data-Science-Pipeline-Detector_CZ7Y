```python
import numpy as np 
import pandas as pd 
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVR
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
df = pd.read_csv('../input/commonlitreadabilityprize/train.csv')
df.excerpt = df.excerpt.str.lower()
df.sample(10)
```

```python
vectorizer = TfidfVectorizer(analyzer='char',ngram_range=(1,6))
X = vectorizer.fit_transform(df.excerpt,)
X.shape
```

```python
model = LinearSVR(random_state=42)
```

```python
scores = cross_val_score(model, X, df.target, cv=5,scoring='neg_root_mean_squared_error')
scores *=-1
scores.mean()
```

```python
model.fit(X,df.target)
```

```python
sub = pd.read_csv('../input/commonlitreadabilityprize/test.csv',index_col='id')
x = vectorizer.transform(sub.excerpt)
p = model.predict(x)
sub['target'] = p
sub[['target']].to_csv('submission.csv')
```
