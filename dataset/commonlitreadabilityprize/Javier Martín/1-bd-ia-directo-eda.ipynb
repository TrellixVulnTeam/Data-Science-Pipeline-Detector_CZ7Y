# Ejercicios
## ¿Qué tipos distinto de extensión tienen las fuentes?
Sugerencia: usar `apply` para sacar la extensión fila a fila


```python
def extension(row):
    f = row.url_legal.split('/')[-1]
    if '.' in f:
        return f.split('.')[-1]
    else:
        return ''

df[df.url_legal.notna()].apply(extension, axis=1).unique()
```

## ¿Cuántos pasajes se han sacado de cada fuente? 
Pista: sería como hacer un SELECT COUNT(excerpt) ... GROUP BY url_legal

```python
pasajes_por_fuente = df.groupby('url_legal').excerpt.count()
pasajes_por_fuente
```
