```python
X = df['avg_word_len'].values
X = X.reshape(-1, 1)
X.shape
```

```python
Y = df['target'].values
Y
```

```python
# ajusta los datos a la ecuación lineal y=ax+b
reg = LinearRegression().fit(X, Y)
```

# Entrenamiento con 5 folds y predicciones out-of-fold (OOF)

```python
from sklearn.model_selection import KFold
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=33)
split = list(kf.split(X))
```

```python
len(split[0][0]), len(split[0][1])
```

```python
for train_index, valid_index in kf.split(X):
    Xt = X[train_index].reshape(-1, 1)
    Yt = Y[train_index]
    reg = LinearRegression().fit(Xt, Yt)
    a, b = reg.coef_[0], reg.intercept_
    Xv = X[valid_index]
    
    df.loc[valid_index, 'pred'] = a * Xv + b
    
```

```python
df
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(df['target'], df['pred'], squared=False)
```

```python
def flesch_re(row):
    flr = textstat.flesch_reading_ease(row['excerpt'])
    return flr

df['flesch'] = df.apply(flesch_re, axis=1)
```
