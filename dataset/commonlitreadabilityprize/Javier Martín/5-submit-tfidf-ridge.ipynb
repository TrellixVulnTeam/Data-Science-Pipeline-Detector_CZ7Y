# Ejercicio
- Predecir con los 5 folds y hacer el promedio

# Solución

```python
from pathlib import Path
```

```python
preds = []
for model_p in Path('/kaggle/input/4-regresion-tf-idf/model/').glob('*.joblib'):
    model = load(model_p)
    preds.append(model.predict(X))
```

```python
preds = np.stack(preds)
preds.shape
```

```python
mean_preds = np.mean(preds, axis=0)
mean_preds.shape
```

```python
mean_preds
```

```python
test_df['target'] = mean_preds
```

```python
test_df
```

```python
test_df = test_df[['id', 'target']]
test_df
```

```python
test_df.to_csv('submission.csv', index=False)
```
