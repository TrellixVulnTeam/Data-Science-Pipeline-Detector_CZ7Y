```python
import textstat
```

```python
test_df = pd.read_csv('/kaggle/input/commonlitreadabilityprize/test.csv')
```

```python
test_df
```

```python
test_df.drop(['url_legal', 'license'], axis=1, inplace=True)
test_df
```

```python
def flesch_re(row):
    flr = textstat.flesch_reading_ease(row['excerpt'])
    return flr

test_df['flesch'] = test_df.apply(flesch_re, axis=1)
```

```python
test_df
```

```python
a, b = (0.030565631533022518, -3.0298994698883406)
```

```python
test_df['target'] = a * test_df['flesch'] + b
```

```python
test_df
```

```python
test_df = test_df[['id', 'target']]
```

```python
test_df
```

```python
test_df.to_csv('submission.csv', index=False)
```
