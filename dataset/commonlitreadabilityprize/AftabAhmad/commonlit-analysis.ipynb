### User define function to remove unwanted text patterns

```python
def replace_pattern(input_text, pattern):
    r = re.findall(input_text,pattern)
    for i  in r:
        input_text = re.sub(i,'',pattern)
    return input_text
```

### Remove special character and puncuation

```python
# Remove special character and puncuation
train['tidy_excerpt'] = train['excerpt'].str.replace('[^a-zA-Z]',' ')
train.head()
```

### Remove words having lenght < 3 or equal to 3

```python
train['tidy_excerpt'] = train['tidy_excerpt'].apply(lambda x:' '.join([w for w in x.split() if len(w)>3]))
train['tidy_excerpt'].head()
```

### Tekenized

```python
#Tekenized
Tokenized_words = train['tidy_excerpt'].apply(lambda x: x.split())
Tokenized_words.head()
```

### Stemming

```python
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()
Tokenized_words=Tokenized_words.apply(lambda x:[stemmer.stem(i) for i in x])
Tokenized_words.head()
```

### Stitching back

```python
#Stitching back
for i in range(len(Tokenized_words)):
    Tokenized_words[i] = ' '.join(Tokenized_words[i])

train['tidy_excerpt'] = Tokenized_words

train.head()

```
