```python
import pandas as pd

import seaborn as sns

import matplotlib.pyplot as plt

from urllib.parse import urlsplit
pd.set_option('display.max_colwidth', 1000)
```

### Reading the data

Let's load only `url_legal` and `target` in this experiment.

```python
train = pd.read_csv('../input/commonlitreadabilityprize/train.csv', usecols=['url_legal', 'target'])
```

We are going to fill the **NaN** urls only to get a better visualization. :)


```python
train['url_legal'].fillna('http://missing',inplace=True)
```

```python
train.tail(25).style.background_gradient(cmap='viridis')
```

### Getting the base url

```python
train['net_loc'] = train['url_legal'].apply(lambda n : urlsplit(n).netloc)
```

```python
g = sns.catplot(x="net_loc", y="target", data=train, height=5, aspect=4, kind="box")
g.set_xticklabels(rotation=60)
```

> Educational websites seems to have a better score, on the other hand `en.wikipedia.org` is almost always bellow -1
