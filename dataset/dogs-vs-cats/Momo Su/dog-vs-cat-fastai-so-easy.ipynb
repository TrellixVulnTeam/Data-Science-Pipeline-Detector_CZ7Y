# Check Data & Build Dataloaders

```python
path = "./train"
files = get_image_files(path)
```

```python
len(files)
```

```python
files[:5]
```

```python
dls = ImageDataLoaders.from_name_re(
        path=path,
        fnames=files,
        pat=r'(.+)\.\d+.jpg$',
        valid_pct=0.2, 
        seed=13,
        item_tfms=Resize(224))

dls.show_batch()
```

# Build CNN Model & Start Training

```python
learn = cnn_learner(dls, resnet34, pretrained=True, metrics=[accuracy])
```

```python
learn.lr_find()
```

```python
learn.fine_tune(1, 1e-3)
```

```python
learn.lr_find()
```

```python
learn.fine_tune(3, 7e-7)
```

# Evaluate Model

```python
interp = ClassificationInterpretation.from_learner(learn)
```

```python
interp.plot_confusion_matrix()
```

```python
interp.plot_top_losses(9, figsize=(9,9))
```
