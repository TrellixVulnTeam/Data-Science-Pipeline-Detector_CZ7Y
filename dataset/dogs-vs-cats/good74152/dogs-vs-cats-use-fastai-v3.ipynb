```python
fnames = get_image_files(trainpath)
```

```python
fnames[:5]
```

```python
labels = [('cat' if 'cat' in str(x) else 'dog') for x in fnames]
```

```python
labels[:5]
```

```python
data = ImageDataBunch.from_lists(trainpath, fnames, ds_tfms=get_transforms(), size=224, bs=64, labels = labels)
data.classes
```

為何size是224?:因為最後一層是7*7 224比較好

Dogs vs cats的特色 : 
1. label為檔名
2. 兩種類別位在同個資料夾

```python
data.show_batch(rows=3, figsize=(5,5))
```

秀出圖片及其label

```python
learn = cnn_learner(data, models.resnet34, metrics=error_rate, model_dir="/tmp/")
```

在kaggle上需要創建一個tmp儲存模型, 不然無法操作模型

```python
learn.fit_one_cycle(4)
```

使用fit_one_cycle訓練

```python
learn.lr_find()
learn.recorder.plot()
```

可視化lr&loss 找學習率的方法:看哪一點開始loss往下降且是平滑陡坡 就是好學習率

```python
interp = ClassificationInterpretation.from_learner(learn)
```

```python
interp.plot_confusion_matrix(figsize=(12,12), dpi=60)
```
