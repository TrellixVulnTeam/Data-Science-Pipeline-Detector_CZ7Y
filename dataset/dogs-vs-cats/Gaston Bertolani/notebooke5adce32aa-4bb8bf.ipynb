```python
import matplotlib.pyplot as plt
def plot_loss_history(history):     
  hist = pd.DataFrame(history.history)
  hist['epoch'] = history.epoch

  plt.figure()
  plt.xlabel('Epoch')
  plt.ylabel('Loss')
  plt.plot(hist['epoch'], hist['loss'],'r--',
           label='Training Error')
  plt.plot(hist['epoch'], hist['val_loss'],'b',
           label = 'Validation Error')
  plt.ylim([0,1])
  plt.legend()
  plt.show()
    
def plot_acc_history(history):     
  hist = pd.DataFrame(history.history)
  hist['epoch'] = history.epoch

  plt.figure()
  plt.xlabel('Epoch')
  plt.ylabel('Accuracy')
  plt.plot(hist['epoch'], hist['accuracy'],'r--',
           label='Training Error')
  plt.plot(hist['epoch'], hist['val_accuracy'],'b',
           label = 'Validation Error')
  plt.ylim([0,1])
  plt.legend()
  plt.show()
```

```python
TRAIN_PATH = './training/'
TEST_PATH = './testing'





traindata = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)

testdata = ImageDataGenerator(rescale = 1./255)

training_set = traindata.flow_from_directory(TRAIN_PATH,
                                                target_size = (150, 150),
                                                batch_size = 20,
                                                class_mode = 'binary')

test_set = testdata.flow_from_directory(TEST_PATH,
                                            target_size = (150, 150),
                                            batch_size = 20,
                                            class_mode = 'binary')
```

```python
from tensorflow.keras.applications import VGG16

pre_trained_model = VGG16(input_shape = (150, 150, 3), 
                                include_top = False, 
                                weights = 'imagenet')

for layer in pre_trained_model.layers:
  layer.trainable = False


pre_trained_model.summary()
```

```python
from tensorflow.keras import Model

from tensorflow.keras import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

model = Sequential()
model.add(pre_trained_model)
model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))

from keras.optimizers import SGD
opt = SGD(lr=0.001, momentum=0.9)
model.compile(
    optimizer=opt,
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
batch_size = 20

# Generator.n // batch_size 

steps_per_epoch = training_set.n // batch_size
validation_steps = test_set.n // batch_size

print(steps_per_epoch)
# >> 1120
print(validation_steps)
# >> 129

```

```python


hist = model.fit_generator(
    steps_per_epoch=steps_per_epoch,
    generator=training_set,
    validation_data= test_set,
    validation_steps=validation_steps,
    epochs=80,
)

```
