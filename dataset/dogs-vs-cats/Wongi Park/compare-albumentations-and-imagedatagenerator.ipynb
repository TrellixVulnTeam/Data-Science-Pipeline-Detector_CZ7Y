# Retrieve Image Data and Make Function
   - 이미지 불러오기 및 함수 제작하기

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import cv2
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import albumentations as alb

org_img = cv2.cvtColor(cv2.imread('../input/cat-and-dog/test_set/test_set/cats/cat.4027.jpg'), cv2.COLOR_BGR2RGB)

# for Albumentations
def Alb_Augmentations(org_img = None, label = None, suptitle = 'Albumentation', albs = None):
    img_list = [org_img]
    label_list = ['orginal']
    for _ in range(3):
        aug_img = albs(image = org_img)['image']
        img_list.append(aug_img)
        label_list.append(label)
    fig, ax = plt.subplots(figsize = (22, 5), nrows = 1, ncols = 4)
    for i in range(4):
        ax[i].imshow(img_list[i])
        ax[i].set_title(label_list[i])
    
    plt.suptitle(f'{suptitle}')
    
# ImageDataGenerator
def Idg_Augmentations(image, data_generator, label = None, suptitle = 'ImageDataGenerator'):
    img = np.expand_dims(image, axis = 0)
    data_generator.fit(img)
    img_gen = data_generator.flow(img)
    
    fig, ax = plt.subplots(figsize = (22, 5), nrows = 1, ncols = 4)
    for i in range(4):
        if i == 0:
            ax[i].imshow(image)
            ax[i].set_title('Orginal')
        else:
            img_batch = next(img_gen)
            img_batch = np.squeeze(img_batch).astype('int')
            ax[i].imshow(img_batch)
            ax[i].set_title(label)

    plt.suptitle(f'{suptitle}')
        
```

# Horizontal Flip, Vertical Flip
- Albumentation
    - P : Probabilty (default = 0.5)
        - P : 확률(기본 = 0.5)

```python
# Albumentation
albs = alb.HorizontalFlip(p = 0.5)
Alb_Augmentations(org_img=org_img, label = 'Horizontal', albs = albs)

# ImageDataGenerator
data_generator = ImageDataGenerator(horizontal_flip=True)
Idg_Augmentations(image=org_img, data_generator=data_generator, label = 'Horizontal')
```

```python
# Albumentation
albs = alb.VerticalFlip(p = 0.5)
Alb_Augmentations(org_img=org_img, label = 'Vertical', albs = albs)

# ImageDataGenerator
data_generator = ImageDataGenerator(vertical_flip=True)
Idg_Augmentations(image=org_img, data_generator=data_generator, label = 'Vertical')
```

# Random Rotation
- Albumentation
    - border_mode : Define the Boundaries (경계를 정의합니다.)
        - cv2.BORDER_CONSTANT  : add a border of a certain color (일정한 색상을 추가합니다.)
        - cv2.BORDER_REPLICATE : Dulipcate the last element (일정한 색상을 복제합니다.)
        - cv2.BORDER_REFLECT : add a border like mirrow (거울과 같이 반사된 형태를 추가합니다.)
        - cv2.BORDER_WRAP : add the orginal image to the margin (여백에 원래 이미지를 추가합니다.)
        - cv2.BORDER_REFLECT_101 : There is sligt difference with BORDER_REFLECT 
            - (BORDER_REFLECT 와 미세한 차이가 있습니다.)
- ImageDataGenerator
    - fill_mode : same border_mode (위와 동일합니다.)
        - constant == cv2.BORDER_CONSTANT
        - nearest == cv2.BORDER_REPLICATE
        - reflect == cv2.BORDER_REFLECT
        - wrap == cv2.BORDER_WRAP

```python
# Albumentation
albs = alb.Rotate(limit= 150, p = 0.8, border_mode = cv2.BORDER_CONSTANT)
Alb_Augmentations(org_img=org_img, label = 'Rotation', albs = albs)

# ImageDataGenerator
data_generator = ImageDataGenerator(rotation_range=150 ,fill_mode = 'constant')
Idg_Augmentations(image=org_img, data_generator=data_generator, label = 'Rotation')
```

# Shift and Scale
- Albumentation
    - shift_limit : Image Move (Image가 이동합니다.)
        - if you adapt 0 for shift_limit, Image doesn't move (0을 적용하면 이미지가 움직이지 않습니다.)
    - scale_limit : Image enlargement or reduction(Image 확대 및 축소)
        - if you adapt 0 for scale_limit, Image doesn't move (0을 적용하면 이미지가 확대되지 않습니다.)
        - if you adapt more 1 scale_limit, Image reduction (1이상 적용시 이미지가 축소됩니다.)
    - rotate_limit : Image Rotation (Image 회전)
    
<br> 


- ImageDataGenerator
    - width_shift_range : Image moves left or right (왼쪽 또는 오른쪽으로 움직입니다.)
    - height_shift_range :Image moves up or down (위, 아래로 움직입니다.)
    - zoom_range : equal scale_limit (위의 scale_limit과 동일합니다.)
    
    - if you wan't move, delete width_shift_range, height_shift_range 
        - (만약 이동을 원하지 않는다면, width_shift_range, height_shift_range 제거 해주시면 됩니다.)

```python
# Albumentation
albs = alb.ShiftScaleRotate(shift_limit=(0.4, 0.6), scale_limit= (0.4, 0.9), p = 0.8, rotate_limit = 0)
Alb_Augmentations(org_img=org_img, label = 'Shift and Scale', albs = albs)

# ImageDataGenerator
data_generator = ImageDataGenerator(width_shift_range=0.4, height_shift_range=0.4, zoom_range= [0.4, 1.5])
Idg_Augmentations(image=org_img, data_generator=data_generator, label = 'Shift and Scale')
```

# Crop
- Only albumentations has Crop (Albumentations 만 Crop 기능을 가지고 있습니다.)
    - Crop doesn't resize orginal image. (원본 이미지로 이지미 크기를 변환하지 않습니다.)
    - if you want resize orginal image, use Albumentations.Composer
        - (만약 원본 이미지로 변환하고 싶다면, Albumentations.Composer을 사용해야 합니다.)
        
<br>

- Crop Kinds
    - CenterCrop : Crop image center (not resize orgin image)
        - 이미지의 중심부를 자릅니다. (원본이미지의 크기로 변환하지 않습니다.)
    - RandomCrop : Crop image random (not resize orgin image)
        - 이미지를 무작위로 자릅니다. (원본이미지의 크기로 변환하지 않습니다.)
        
<br>

* if you want resize orginal image, use Albumentations.Composer
    - (만약 원본 이미지로 변환하고 싶다면, Albumentations.Composer을 사용해야 합니다.)

```python
# Albumentation
albs = alb.Crop(x_min = 100, y_min = 100, x_max = 300, y_max = 300, p = 0.8)
Alb_Augmentations(org_img=org_img, label = 'Crop', albs = albs)

albs = alb.Compose([
    alb.Crop(x_min = 100, y_min = 100, x_max = 300, y_max = 300, p = 0.8),
    alb.Resize(621, 934)
])
Alb_Augmentations(org_img=org_img, label = 'Crop', albs = albs)
```

```python
# Albumentation
albs = alb.Compose([
    alb.CenterCrop(height = 100, width = 300, p = 0.8),
    alb.Resize(621, 934)
])
Alb_Augmentations(org_img=org_img, label = 'CenterCrop', albs = albs)
# Albumentation
albs = alb.Compose([
    alb.RandomCrop(height = 100, width = 300, p = 0.8),
    alb.Resize(621, 934)
])
Alb_Augmentations(org_img=org_img, label = 'RandomCrop', albs = albs)
```

# RandomBrightnessContest
- Albumentation
    - brightness_limit : Set a Scope(range) in tuple. (튜플 내에 범위를 정해야 합니다.)
    - contrast_limit : Set a Scope(range) in tuple. (튜플 내에 범위를 정해야 합니다.)

<br>

- ImageDataGenerator
    - brightness_range : Set a Scope(brightness_range) in tuple. (튜플 내에 밝기 범위를 정해야 합니다.)

```python
# Albumentation
albs = alb.RandomBrightnessContrast(brightness_limit=(-0.4, 0.4), contrast_limit=(-0.4, 0.4), p=0.8)
Alb_Augmentations(org_img=org_img, label = 'RandomBrightnessContrast', albs = albs)

# ImageDataGenerator
data_generator = ImageDataGenerator(brightness_range=(0.1, 0.8))
Idg_Augmentations(image=org_img, data_generator=data_generator, label = 'Bright')
```

<br>

- ImageDataGenerator
    - channel_shfit_range : RGB +- value in range (원본 RGB 값에서 범위 내의 값을 +- 합니다)

```python
# Albumentation
albs = alb.ChannelShuffle(p=0.8)
Alb_Augmentations(org_img=org_img, label = 'ChannelShuffle', albs = albs)

# ImageDataGenerator
data_generator = ImageDataGenerator(channel_shift_range = 60 )
Idg_Augmentations(image=org_img, data_generator=data_generator, label = 'channel_shift_range')
```

```python
# Albumentation
albs = alb.CLAHE(p=0.8)
Alb_Augmentations(org_img=org_img, label = 'CLAHE', albs = albs)
```

# Oneof
- adaptive one contents in Oneof (Oneof 내의 한가지를 적용합니다.) 

```python
# Albumentation
albs = alb.OneOf([
    alb.Rotate(limit=(50, 150), p = 1, border_mode= cv2.BORDER_CONSTANT),
    alb.HorizontalFlip(p = 1),
    alb.ChannelShuffle(p = 1),
    alb.RGBShift(p = 1)
])
Alb_Augmentations(org_img=org_img, label = 'Oneof', albs = albs)
```

# Compose
- Albumentations
    - it's similar ImageDataGenerator (ImageDataGenerator 와 비슷하다고 보면 됩니다.)

```python
# Albumentation
albs = alb.Compose([
    alb.Rotate(limit=(50, 150), p = 1, border_mode= cv2.BORDER_CONSTANT),
    alb.HorizontalFlip(p = 1),
    alb.ChannelShuffle(p = 1),
    alb.RGBShift(p = 1),
    alb.CLAHE(p = 1),
    alb.GaussNoise(p = 1, var_limit = (50, 100)),
    alb.CoarseDropout(p = 1, max_holes = 50)
])
Alb_Augmentations(org_img=org_img, label = 'Compose', albs = albs)

# ImageDataGenerator
data_generator = ImageDataGenerator(vertical_flip=True,
                                    horizontal_flip=True,
                                    shear_range=True,
                                    rotation_range= 150,
                                    zoom_range = 0.7
                                   )
Idg_Augmentations(image=org_img, data_generator=data_generator, label = 'ImageDataGenerator')
```
