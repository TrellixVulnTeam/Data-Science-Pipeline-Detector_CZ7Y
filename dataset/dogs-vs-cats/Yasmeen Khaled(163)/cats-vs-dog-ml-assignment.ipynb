```python
#/kaggle/input/dogs-vs-cats/test1.zip

data_set = "dogs-vs-cats"

import zipfile 
with zipfile.ZipFile("/kaggle/input/"+ data_set +"/train.zip","r") as z:
    z.extractall(".")
    # save all files to kaggle/files/images
    destination = '/kaggle/files/images'
    z.extractall(destination)
```

```python
data_ = pd.DataFrame({'file': os.listdir('/kaggle/files/images/train')})
#print(data_.head())
```

```python


Y=[]
for i in os.listdir('/kaggle/files/images/train'):
    if 'dog' in i:
        Y.append(1)
    else:
        Y.append(0)
    
        
data_['class'] = Y
#print(data_.head())
```

```python
file = data_['file']  

Y = data_['class']
```

```python
# importing matplotlib modules
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

from skimage.feature import hog
from skimage import data, exposure
from skimage.transform import rescale, resize


data_size = 1200
ptr=0
X = []
for i in file:
    img = mpimg.imread('/kaggle/files/images/train/' + i)
    resized_img = resize(img , (128 , 64))
    fd, hog_image = hog(resized_img, orientations=9, pixels_per_cell=(8, 8),cells_per_block=(2, 2), visualize=True, multichannel = True)
    X.append(fd)
    ptr = ptr+1
    if(ptr >= data_size):
        break
```

```python
#Y = Y.tolist()
Y = Y[:data_size]

# dividing X, y into train and test data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, random_state=0)
#X_train.head()
#print(type(X_train))#<class 'pandas.core.series.Series'>
```

```python
#print(type(X_train[0])) #<class 'numpy.ndarray'>
#print(X_train[0].shape) #(3780,)

# n is large relative to m (n = (3780,)  m=1200) --> model: SVM without a kernel
```

```python
# svm_LinearSVC  min(h^2)
c = 1

from sklearn.svm import LinearSVC
svm_LinearSVC = LinearSVC(C=c).fit(X_train, y_train) #minimize hinge loss Squared
```

```python
#svm_svc  min(h)

from sklearn.svm import SVC
svm_svc = SVC(kernel='linear', C=c).fit(X_train, y_train)
```

```python
# svm_LinearSVC accuracy
accuracy = svm_LinearSVC.score(X_test, y_test)
print('svm_LinearSVC accuracy: ' + str(accuracy))

#svm _svc accuracy
accuracy = svm_svc.score(X_test, y_test)
print('svm_svc accuracy: ' + str(accuracy))
print("c = ",c)
```
