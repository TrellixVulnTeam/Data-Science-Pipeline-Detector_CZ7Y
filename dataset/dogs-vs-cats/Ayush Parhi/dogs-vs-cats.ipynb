```python
fnames = get_image_files('/kaggle/working/train')
fnames[:5]
```

```python
np.random.seed(2)
pat = r'(dog|cat)'
```

```python
data = ImageDataBunch.from_name_re("/kaggle/working/train", fnames, pat, ds_tfms = get_transforms(), size = 224, bs = 64).normalize(imagenet_stats)
```

```python
data.show_batch(rows = 3, figsize = (7,6))
```

```python
print(data.classes)
```

```python
len(data.classes), data.c
```

```python
learn = cnn_learner(data, models.resnet34, metrics = error_rate)
```

```python
learn.fit_one_cycle(5)
```

```python
learn.save('stage-1')
```

```python
learn.unfreeze()
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.fit_one_cycle(2, max_lr = slice(1e-05, 1e-04))
```

```python
learn.save('stage-2')
```

```python
learn.load('stage-2')
```

```python
interp = ClassificationInterpretation.from_learner(learn)
```

```python


interp.plot_confusion_matrix()
```
