## Data Preprocessing

```python
# İleride işlem yapabilmek için float'a çevirelim. Ve scale edelim. 255 --> 1 ve 0 --> 0
# Let's convert it to float for further processing. And scale them. 

X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
X_train /= 255
X_test /= 255
```

## <span style="color:LightCoral; background-color:#F5EEF8"><font size="4">KERAS to_categorical</font></span> <a id="4.1.1"></a>

Bir sınıf(class) vektörünü (yani tam sayıları) ikili sınıf matrisine dönüştürür. Daha fazlasını öğrenmek istiyorsanız lütfen [tıklayın](https://www.tensorflow.org/api_docs/python/tf/keras/utils/to_categorical).<br>
<span style="color:#1565C0; opacity:1">Converts a class vector (integers) to binary class matrix. If you wanna learn more please </span>[click](https://www.tensorflow.org/api_docs/python/tf/keras/utils/to_categorical)<br>

```python
from tensorflow.keras.utils import to_categorical
```

```python
Y_train = to_categorical(y_train, 10)  # --> y_traini alır ve kategorik hale getirir. Class sayısını 10 olarak ayarladım.
Y_test = to_categorical(y_test, 10)    # --> takes y_test like above and makes it categorical. Number of classes is 10.
Y_test.shape
```

```python
from tensorflow.keras.models import Sequential  # Sequential model.
from tensorflow.keras.layers import Dense, Conv2D, MaxPool2D, Flatten

# 1 --> Model
# 2 --> conv. layer, 
# 3 --> maxpool, 
# 4 --> padding
# 5 --> flatten layer
# 6 --> Hidden layer and it's activation function
# 7 --> Output layer and it's activation function / dense layer
```

```python
# 1) Model seçimi / Model selection
model = Sequential()

# 2) FILTER - KERNEL
# 32 times --> 3X3 kernel(filter), activation --> relu ve input --> 1X28X28
model.add(Conv2D(filters=32, kernel_size=(3,3),input_shape=(28, 28, 1), activation='relu',))  

# 3) POOLING
# 28 ve 32 lik boyutlar için 2x2 lik pooling yeterli. (For sizes 28 and 32, 2x2 pooling is fairly enough.) 
model.add(MaxPool2D(pool_size=(2, 2)))  

# 4) PADDING
# Padding kurmadık çünkü köşelerde veri yok, olsaydı parametreyi same ayarlarık. (No Padding --> no data at the edges) 
# stride --> 2 (default)

# 5) FLATTEN 
# matris --> vektöre dönüşümü / matris --> vector transformation
model.add(Flatten())

# 6) Activation Function
# nöron sayısı(quantity) --> 128  ve activation --> relu
model.add(Dense(128, activation='relu'))

# 7) ANN deki gibi 2. hidden layer isteğe bağlı kuruldu.
# As in ANN, the 2nd hidden layer was optionally setted.
model.add(Dense(10, activation='softmax'))

model.compile(loss='categorical_crossentropy',
              optimizer='adam',  # OPTIMIZER !!!
              metrics=['accuracy'])
```

## <span style="color:LightCoral; background-color:#F5EEF8"><font size="4">Earlystopping</font></span> <a id="4.1.2"></a>

İzlenen bir metrik iyileştirmeyi durdurduğunda eğitimi durdurur. Daha fazlasını öğrenmek istiyorsanız lütfen [tıklayın](https://www.tensorflow.org/api_docs/python/tf/keras/callbacks/EarlyStopping).<br>
Eğer görsel ve uygulamalı olarak izleyerek öğrenmek isterseniz lütfen videoya [buradan](https://www.youtube.com/watch?v=2UHCjhyNLKw) erişin.<br>
<span style="color:#1565C0; opacity:1">It stops training when a monitored metric has stopped improving. If you wanna learn more please</span> [click](https://www.tensorflow.org/api_docs/python/tf/keras/callbacks/EarlyStopping)<br>

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
# Düşüşünü yani öğrenmesini takip etmemiz gereken değeri monitor değişkenine atarız
# We assign the value we need to follow to the variable "monitor".

early_stop = EarlyStopping(monitor='val_loss',patience=2)
```

## <span style="color:LightCoral; background-color:#F5EEF8"><font size="4">Batch Size / Epochs / Learning Rate</font></span> <a id="4.1.3"></a>

İterasyonlar yani tekrarlar, bir epoch'u tamamlamak için gereken batch sayısıdır. Batch sayısı, bir epoch içindeki iterasyon sayısına eşittir.<br>
Kısaca: Bizim datamız 60000 satır idi. Aşağıda her bir batch içinde 1875 satır var ve batch sayımız 32'dir. 1875x32=60000<br>
Ve her bir epoch'ta hata kayıbımız global point'e düşmesi için eğitimler tekrarlanır.<br>
Hatanın en düşük olacağı global point'i yakalayabilmek için püf noktalar (learning rate etc.) [buradadır](https://towardsdatascience.com/estimating-optimal-learning-rate-for-a-deep-neural-network-ce32f2556ce0).<br>
<span style="color:#1565C0; opacity:1">Iterations is the number of batches needed to complete one epoch.The number of batches is equal to number of iterations for one epoch.<br>
In short: Our data was 60000 rows. Below are 1875 rows in each batch and our batch number is 32. 1875x32=60000.<br>
And in each epoch, the trainings are repeated so that our error loss falls to the global point.<br>
Tricks to catch the global point where the error will be lowest (learning rate etc.)</span> [here](https://towardsdatascience.com/estimating-optimal-learning-rate-for-a-deep-neural-network-ce32f2556ce0)

```python
model.fit(X_train, Y_train, batch_size=32, epochs=10, validation_data=(X_test,Y_test),callbacks=[early_stop])
```

```python
model.metrics_names
```

```python
model.history.history
```

```python
summary = pd.DataFrame(model.history.history)
summary.head()
```

```python
plt.figure(figsize=(10,6))
plt.plot(summary.loss)
plt.plot(summary.val_loss)
plt.ylabel("Loss")
plt.xlabel("Epoch")
plt.gca().legend(('loss',"val_loss"));
```

```python
print(model.evaluate(X_test,Y_test,verbose=1))  # soldaki loss sağdaki accuracy / left one is loss and the other accuracy
```

Bir klasifikayson modelini değerlendirmek için kullanılan iki teknik, confusion matrisi ve klasifikasyon raporudur.<br>
Daha fazlası için [makale](https://medium.com/swlh/confusion-matrix-and-classification-report-88105288d48f) ve [video](https://www.youtube.com/watch?v=O-eH80oi2P4).<br>
<span style="color:#1565C0; opacity:1">In machine learning, classification is part of supervised learning, which means that the data used to train the model have labels that identify each category.<br>
A critical step in the life cycle of a machine learning model is the evaluation of its performance.<br>
Two techniques used to evaluate a classification model are the confusion matrix and the classification report.<br>
To learn more here is the </span>[article](https://medium.com/swlh/confusion-matrix-and-classification-report-88105288d48f) <span style="color:#1565C0; opacity:1">and</span> [video](https://www.youtube.com/watch?v=O-eH80oi2P4)</span>

**Görüntü Manipülasyonu**

Görüntüleri döndürme, yeniden boyutlandırma ve ölçekleme ile işlemek için ImageDataGenerator'ı kullanabiliriz, böylece model, veri setimizin sahip olmadığı farklı görüntülere karşı daha sağlam hale gelir. ImageDataGenerator aşağıdakileri yapar.


* Eğitim için kullanılan batch görüntüleri kabul eder.
* Batch/grup içindeki her görüntüye bir dizi rastgele dönüşüm uygular.
* Orijinal batch/grup ile rastgele dönüştürülmüş grup resmi değiştirir.
* CNN'yi bu rastgele dönüştürülmüş batchler ile eğitir.


Veri büyütmeyi uygulamanın amacı, daha genelleştirilmiş bir modele sahip olmaktır.

<span style="color:#1565C0; opacity:1;font-weight:bold">Image Manipulation</span>

<span style="color:#1565C0; opacity:1"><br>
We can use the ImageDataGenerator to manipulate the images with rotation, resizing, and scaling so the model becomes more robust to different images that our data set doesn't have. ImageDataGenerator does the followings.</span>

<span style="color:#1565C0; opacity:1"><br>
     *Accepts a batch of images used for training.<br>
     *Applies a series of random transformations to each image in the batch.<br>
     *Replaces the original batch with randomly transformed batch.<br>
     *Training the CNN on this randomly transformed batch.</span>

<span style="color:#1565C0; opacity:1"><br>
The goal of applying data augmentation is to have a more generalized model.</span>

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
image_gen = ImageDataGenerator(
              rotation_range=20,       # Resmi 20 derece döndürür / rotate the image 20 degrees
              width_shift_range=0.10,  # Resim eninin maksimum %5 kaydırır / Shift the pic width by a max of 5%
              height_shift_range=0.10, # Resim boyunu maksimum %5 kaydırır / Shift the pic height by a max of 5%
              rescale=1/255,           # Resmi normalleştirerek yeniden ölçeklendirir / Rescale the image by normalzing it.
              shear_range=0.1,         # Görüntünün bir kısmını keser max 0.1 / Shear means cutting away part of the image (max 10%)
              zoom_range=0.1,          # Maksimum %10 yakınlaştır / Zoom in by 10% max
              horizontal_flip=True,    # Yatay çevirir / Allow horizontal flipping
              fill_mode='nearest'      # Eksik pikselleri en yakındaki ile doldurur / Fill in missing pixels with the nearest filled value
              )
```
