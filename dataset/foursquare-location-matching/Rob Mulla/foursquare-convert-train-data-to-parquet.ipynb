# Parquet Files are Faster to Load
Lets make a parquet version of this data

```python
import pandas as pd
from glob import glob
```

```python
!ls -GFlash ../input/foursquare-location-matching/
```

```python
csv_files = glob('../input/foursquare-location-matching/*.csv')
```

```python
for f in csv_files:
    df = pd.read_csv(f)
    out_fn = f.split('/')[-1].replace('.csv','.parquet')
    df.to_parquet(out_fn)
```
