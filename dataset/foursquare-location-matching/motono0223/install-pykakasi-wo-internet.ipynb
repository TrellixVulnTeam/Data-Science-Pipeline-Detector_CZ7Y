```python
import pykakasi
kakasi = pykakasi.kakasi()
kakasi.setMode('H', 'a') # Convert Hiragana into alphabet
kakasi.setMode('K', 'a') # Convert Katakana into alphabet
kakasi.setMode('J', 'a') # Convert Kanji into alphabet
conversion = kakasi.getConverter()
```

```python
from unidecode import unidecode
```

```python
import pandas as pd
def convert_japanese_alphabet(df: pd.DataFrame):
    def convert(row):
        for column in ["name", "address", "city", "state"]:
            try:
                row[column] = conversion.do(row[column])
            except:
                pass
        return row

    df[df["country"] == "JP"] = df[df["country"] == "JP"].apply(convert, axis=1)
    return df

def convert_unidecode_alphabet(df: pd.DataFrame):
    def convert(row):
        for column in ["name", "address", "city", "state"]:
            try:
                row[column] = unidecode(row[column])
            except:
                pass
        return row

    df = df.apply(convert, axis=1)
    return df
```

```python
df = pd.read_csv("../input/foursquare-location-matching/train.csv").reset_index()
```

```python
df2 = convert_japanese_alphabet( df.iloc[:30] )
df2 = convert_unidecode_alphabet( df2 )
```
