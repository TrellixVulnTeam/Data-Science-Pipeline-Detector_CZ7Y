```python
import pandas as pd

train=pd.read_csv('../input/foursquare-location-matching/train.csv')
test=pd.read_csv('../input/foursquare-location-matching/test.csv')
sample_submission=pd.read_csv('../input/foursquare-location-matching/sample_submission.csv')
pairs=pd.read_csv('../input/foursquare-location-matching/pairs.csv')
```

# 2. Matching checker

Please customize this function as you like.

```python
def matching_check(r1,r2):
    if (r1[3]==r2[3]) and (r1[4]==r2[4]):
        matching_result=True
    else:
        matching_result=False
    return matching_result    
```

# 3. Simple loop processing

```python
sub={'id':[],'matches':[]}
for row1 in test.itertuples():
    sub['id'].append(row1[1])
    match=''
    for row2 in test.itertuples():
        if (matching_check(row1,row2)==True):
            match=match+' '+row2[1]
    sub['matches'].append(match.lstrip())
```

# 4. Submission

```python
submit=pd.DataFrame(sub)
submit.to_csv("submission.csv",index=False)
submit.head()
```
