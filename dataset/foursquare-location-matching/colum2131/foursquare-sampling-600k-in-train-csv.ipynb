In this notebook, the data size of train.csv is to extract 600k as in test.  
The same POI can all be included in the data by using GroupKFold.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import GroupKFold

train = pd.read_csv('../input/foursquare-location-matching/train.csv')
display(train)
```

```python
folds = pd.Series(dtype=int)

kf = GroupKFold(n_splits=11)
for fold, (_, valid_index) in enumerate(kf.split(train, train['id'], train['point_of_interest'])):
    folds = pd.concat([folds, pd.Series(fold, index=valid_index)])
folds = folds.sort_index()  
```

```python
new_train = train[folds<=5].reset_index(drop=True)
display(new_train)
```
