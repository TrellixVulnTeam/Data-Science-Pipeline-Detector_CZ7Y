```python
print('Unique POIs : {}'.format(train.point_of_interest.nunique())) 
```

```python
train.point_of_interest.value_counts()
```

```python
name = train.point_of_interest.value_counts().to_frame().reset_index().rename(columns = {'index':'name', 'point_of_interest':'count'})

fig, ax = plt.subplots()
ax = sns.barplot(data = name.head(20), x = 'name', y = 'count')
ax.set_ylabel('No of Records')
ax.set_xlabel('Name')
plt.xticks(rotation = 90)
plt.show()
```

```python
print('Countries Encompassed : {}'.format(train.country.nunique())) 
```

```python
name = train.country.value_counts().to_frame().reset_index().rename(columns = {'index':'name', 'country':'count'})

fig, ax = plt.subplots()
ax = sns.barplot(data = name.head(20), x = 'name', y = 'count')
ax.set_ylabel('No of Records')
ax.set_xlabel('Name')
plt.xticks(rotation = 90)
plt.show()
```

```python
print('Unique Categories : {}'.format(train.country.nunique())) 
```

```python
train.categories.value_counts().to_frame().reset_index()
```

```python
name = train.categories.value_counts().to_frame().reset_index().rename(columns = {'index':'name', 'categories':'count'})

fig, ax = plt.subplots()
ax = sns.barplot(data = name.head(20), x = 'name', y = 'count')
ax.set_ylabel('No of Records')
ax.set_xlabel('Name')
plt.xticks(rotation = 90)
plt.show()
```

```python
name = train.address.value_counts().to_frame().reset_index().rename(columns = {'index':'name', 'address':'count'})

fig, ax = plt.subplots()
ax = sns.barplot(data = name.head(20), x = 'name', y = 'count')
ax.set_ylabel('No of Records')
ax.set_xlabel('Name')
plt.xticks(rotation = 90)
plt.show()
```
