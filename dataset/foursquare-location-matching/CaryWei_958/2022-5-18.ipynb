```python
#观察数据和缺失数量决定去除 ‘url’, 'phone', 'zip' 列
Train = Train.drop(columns = ['zip', 'url', 'phone'])
Train.info()
```

```python
# 使用knn预测缺失值
from sklearn import neighbors
from sklearn import datasets
knn = neighbors.KNeighborsClassifier(n_neighbors = 2)
# knn 预处理
Data4Knn = Train[['id', 'latitude', 'longitude']].copy()
Y4Knn = Data4Knn[['id']].copy()
X4Knn = Data4Knn[['latitude', 'longitude']].copy()
#训练knn模型
knn.fit(X4Knn,Y4Knn)
```
