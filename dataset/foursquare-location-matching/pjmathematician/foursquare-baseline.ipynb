```python
import pandas as pd
```

```python
test = pd.read_csv("../input/foursquare-location-matching/test.csv")
test.head()
```

```python
submission = {'id':[],'matches':[]}
for i,r in test.iterrows():
    submission['id'].append(r['id'])
    submission['matches'].append(r['id'])
```

```python
sub = pd.DataFrame(submission)
sub.to_csv("submission.csv",index=False)
sub.head()
```
