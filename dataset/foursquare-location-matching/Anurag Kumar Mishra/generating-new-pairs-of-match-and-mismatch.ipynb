We have `id` as row unique identifier and `point_of_interest` as the column we should be interested in. We have been given `pairs` dataset containing the data row having common `point_of_interest`. 
Here we intent to generate more such pairs and also mismatched pairs.

```python
new_pairs_train_merged = pd.merge(train_data, train_data, on='point_of_interest', suffixes=('_1', '_2'), how='inner')
new_pairs_train = new_pairs_train_merged[new_pairs_train_merged['id_1'] != new_pairs_train_merged['id_2']]
new_pairs_train = new_pairs_train.drop(['point_of_interest'], axis=1)
new_pairs_train['match'] = True
```

Now, generating new mismatched. Note that: Mismatched can be generally very high in number, here just providing a method to generate approx equal new mismatch. We shift index and concat and then eliminate common ones. Note you can shuffle data on avoid the linearly placed `point_of_interest`. 

```python
train_data_copy = train_data.copy()
train_data_copy.index = range(1,len(train_data_copy)+1)
train_data_copy = train_data_copy.add_suffix('_2')
non_pairs = pd.concat([train_data.add_suffix('_1'),train_data_copy], axis=1).dropna(subset=['point_of_interest_1', 'point_of_interest_2'])
non_pairs = non_pairs[non_pairs['point_of_interest_1'] != non_pairs['point_of_interest_2']]
non_pairs = non_pairs.drop(['point_of_interest_1', 'point_of_interest_2'], axis=1)
non_pairs['match'] = False
```
