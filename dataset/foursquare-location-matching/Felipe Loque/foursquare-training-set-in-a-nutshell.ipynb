```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("../input/foursquare-location-matching/train.csv", usecols=['latitude','longitude','country'])
```

```python
fig = px.scatter_geo(df[:50000], lat='latitude', lon='longitude', hover_name='country')
fig.update_layout(title='Foursquare Training Set (First 50k)', title_x=0.5)
fig.update_traces(marker=dict(size=2))
fig.update_geos(visible=True, showcountries=True)
fig.show()
```

```python
fig = px.scatter_geo(df[-50000:], lat='latitude', lon='longitude', hover_name='country')
fig.update_layout(title='Foursquare Training Set (Last 50k)', title_x=0.5)
fig.update_traces(marker=dict(size=2))
fig.update_geos(visible=True, showcountries=True)
fig.show()
```
