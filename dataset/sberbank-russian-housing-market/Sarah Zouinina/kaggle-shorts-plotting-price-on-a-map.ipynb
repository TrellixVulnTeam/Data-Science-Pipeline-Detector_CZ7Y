```python
df = pd.read_csv(ZipFile("../input/sberbank-russian-housing-market/train.csv.zip").open('train.csv'), parse_dates=['timestamp'])
df.head()
```

```python
df_geo = pd.read_csv("/kaggle/input/gps-sub-df/gps_sub_area_data.csv")
df_geo.head()
```

```python
new_df = pd.merge(df, df_geo, how="left", on=["sub_area"])
```

```python
import plotly.express as px
px.set_mapbox_access_token("Your Token")

fig = px.scatter_mapbox(new_df, lon="latitude", lat="longitude", size="price_doc",
                  color_continuous_scale=px.colors.cyclical.IceFire, size_max=15, zoom=10, width=500, height=800)
#fig.show()
```
