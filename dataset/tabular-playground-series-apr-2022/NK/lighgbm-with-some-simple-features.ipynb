## preprocess

```python
df_train_3.columns
```

```python
x_train = df_train_3[df_train_3["step"]==0].iloc[:,17:]
x_train
```

```python
### fix

y_train = df_train_3[df_train_3["step"]==0].iloc[:,16]
y_train
```

## lightgbm

```python
import lightgbm as lgb
from sklearn.model_selection import KFold

paras = {
    
        'boosting_type': 'gbdt',
        'objective': 'binary',
        'learning_rate': 0.1,
        'max_depth': 5,
        'random_state': 2022,
        'bagging_fraction': 0.8,
        'feature_fraction': 0.8, 
        'metric': 'auc'
        }

kf = KFold(n_splits = 5, shuffle = True, random_state = 70)
models = []
for fold, (tr_idx, va_idx) in enumerate(kf.split(x_train)):
    print(f'--------fold:{fold+1}--------')
    fold+=1
    tr_x, va_x = x_train.iloc[tr_idx], x_train.iloc[va_idx]
    tr_y, va_y = y_train.iloc[tr_idx], y_train.iloc[va_idx]
    data_train = lgb.Dataset(tr_x, tr_y)
    data_val = lgb.Dataset(va_x, va_y)
    
    
    lgb_results = {}   
    model = lgb.train(
        params = paras,
        train_set = data_train,
        valid_sets = [data_val ,data_train],
        valid_names=['eval', 'train'],
        num_boost_round = 1000,
      #  valid_sets = watchlist,
        early_stopping_rounds=50,
        evals_result=lgb_results,
        verbose_eval=100
    )
    models.append(model)
```

# make csv for submit

```python
df_test_4 = df_test_3[df_test_3["step"]==0]
df_test_4.columns
```

```python
x_test = df_test_4.iloc[:,16:]
x_test
```

```python
prediction = np.zeros(x_test.shape[0])
for i,model in enumerate(models):
    pred = model.predict(x_test) 
    prediction += pred
prediction = prediction/len(models)
```

```python
fig = plt.figure(figsize=(16,12))

ax1 = fig.add_subplot(221)
ax1.set_xlabel('predict')
ax1.set_ylabel('count')
ax1.hist(prediction,label='test',bins=50)

ax2 = fig.add_subplot(222)
ax2.set_xlabel('predict')
ax2.set_ylabel('count')
ax2.hist(prediction, label='test',cumulative=True,bins=50)

plt.legend()
plt.show()
```
