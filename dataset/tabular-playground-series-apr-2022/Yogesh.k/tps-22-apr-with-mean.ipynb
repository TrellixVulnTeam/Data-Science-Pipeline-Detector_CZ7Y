```python
df_test = pd.DataFrame(test.groupby('sequence')['sensor_00', 'sensor_01', 'sensor_02',
       'sensor_03', 'sensor_04', 'sensor_05', 'sensor_06', 'sensor_07',
       'sensor_08', 'sensor_09', 'sensor_10', 'sensor_11', 'sensor_12'].agg('mean'))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model1 = RandomForestClassifier()
model1.fit(df,train_lables['state'])
```

```python
pred1 = model1.predict(df_test)
```

```python
sample_submission['state']=pred1
sample_submission.to_csv('ss.csv')
```

```python
from sklearn.metrics import classification_report, accuracy_score, plot_confusion_matrix
```

```python
pred_train = model1.predict(df)
```

```python
accuracy_score(train_lables['state'],pred_train)
```

```python
plot_confusion_matrix(model1,df,train_lables['state'])
```
