As noted by @ambrosm, the output probability highly depend on the number of subject occurence. 
I leverage this to produce a no ML baseline. I am not sure this feature would be so usefull in real life.
(Oversampling ill people because they often come to the hospital wouldn't be very helpfull to diagnose new patient out of the blue).

```python
import pandas as pd

path = '../input/tabular-playground-series-apr-2022/'

df = pd.read_csv(path+'test.csv')
submission = pd.read_csv(path+'sample_submission.csv')

map_sequence_subject = df.groupby(['sequence']).subject.min()
map_subject_count = map_sequence_subject.value_counts()
submission['state'] = submission['sequence'].map(map_sequence_subject.map(map_subject_count))

submission.to_csv('submission.csv', index=False)
```
