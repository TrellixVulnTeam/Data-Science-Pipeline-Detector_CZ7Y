```python
threshold_l = 0.01
threshold_u = 0.99
```

# We weight the data by their public score

```python
lstm1_weight = 0.973
lstm2_weight = 0.977
xgb_weight = 0.947
xgb_bru_weight = 0.971

cnn_weight = 0.971
```

```python
weighted_mean = (xgb_weight* xgb["state"]+ xgb_bru_weight* xgb_bru["state"] + 
                 lstm1_weight* lstm1["state"] + lstm2_weight* lstm2["state"] +
                 cnn_weight* cnn["state"]) / (lstm1_weight +lstm2_weight+ xgb_weight +xgb_bru_weight + cnn_weight)
sub["state"]= sub["state"] = weighted_mean.values
sub
```

```python
sub_no_weight = sub.copy(deep = True)
sub_no_weight["state"]= ((xgb["state"]+xgb_bru["state"]++lstm1["state"] +lstm2["state"]+ cnn["state"])/5).values 
sub_no_weight.to_csv("sub_no_weight.csv")
```

```python
sub_modified = sub.copy(deep = True)
print(f"Over {threshold_u} probability:" ,len(sub_modified.loc[sub["state"] >=threshold_u,"state"]))
print(f"Less than {threshold_l} probability:" ,len(sub_modified.loc[sub["state"] <=threshold_l,"state"]))
sub_modified.loc[sub["state"] >=threshold_u,"state"] = 1
sub_modified.loc[sub["state"] <=threshold_l,"state"] = 0
```

```python
plt.figure(figsize = (20,8))
sns.histplot(sub["state"],color = "b" , alpha = 0.5, label = "As is")
sns.histplot(sub_modified["state"],color = "red" , alpha = 0.5,label = "modified")
plt.legend()
plt.show()
```
