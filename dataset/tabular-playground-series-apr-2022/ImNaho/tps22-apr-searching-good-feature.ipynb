# read data

```python
train = pd.read_csv('/kaggle/input/tabular-playground-series-apr-2022/train.csv')
train_label = pd.read_csv('/kaggle/input/tabular-playground-series-apr-2022/train_labels.csv')
test = pd.read_csv('/kaggle/input/tabular-playground-series-apr-2022/test.csv')
```

# check kdeplot

```python
cont_features = train.columns
print("Feature distribution of important columns : ")
ncols = 3
nrows = 4

fig, axes = plt.subplots(nrows, ncols, figsize=(40,15), facecolor='#EAEAF2')

for r in range(nrows):
    for c in range(ncols):
        col = cont_features[r*ncols+c+3]
        sns.kdeplot(x=train[col], ax=axes[r, c], color='red', label='Train data')
        sns.kdeplot(x=test[col], ax=axes[r, c], color='blue', label='Test data')
        axes[r, c].set_ylabel('')
        axes[r, c].set_xlabel(col, fontsize=15, fontweight='bold')
        axes[r, c].tick_params(labelsize=10, width=0.5)
        axes[r, c].xaxis.offsetText.set_fontsize(20)
        axes[r, c].yaxis.offsetText.set_fontsize(20)
plt.show()
```

# line plot with step axis

```python
train_step_=train_step.drop(['sequence','subject'],axis=1)
test_step_=test_step.drop(['sequence','subject'],axis=1)


print("Feature distribution of important columns : ")
ncols = 3
nrows = 4

fig, axes = plt.subplots(nrows, ncols, figsize=(40,20), facecolor='#EAEAF2')

for r in range(nrows):
    for c in range(ncols):
        col = cont_features[r*ncols+c+3]
        sns.lineplot(data=train_step_[col],ax=axes[r, c], x=train_step_.index,y=train_step_[col], color='red', label='Train data')
        sns.lineplot(data=test_step_[col],ax=axes[r, c], x=test_step_.index,y=test_step_[col], color='blue', label='Test data')
plt.show()
```

# are there features depends on states?

```python
sequence_01=train_label.loc[train_label['state']==1,'sequence']
train['state']=0
for sequence_ in sequence_01:
    train.loc[train['sequence']==sequence_,'state']=1

train_sequence=train.groupby('sequence').mean()
train_subject=train.groupby('subject').mean()
```

```python
train_sequence_00=train_sequence[train_sequence['state']==0]
train_sequence_01=train_sequence[train_sequence['state']==1]
```

```python
print("Feature distribution of important columns : ")
ncols = 3
nrows = 4

fig, axes = plt.subplots(nrows, ncols, figsize=(40,15), facecolor='#EAEAF2')

for r in range(nrows):
    for c in range(ncols):
        col = cont_features[r*ncols+c+3]
        sns.kdeplot(x=train_sequence_00[col], ax=axes[r, c], color='red', label='Train 00 data')
        sns.kdeplot(x=train_sequence_01[col], ax=axes[r, c], color='blue', label='Train 01 data')
        axes[r, c].set_ylabel('')
        axes[r, c].set_xlabel(col, fontsize=15, fontweight='bold')
        axes[r, c].tick_params(labelsize=10, width=0.5)
        axes[r, c].xaxis.offsetText.set_fontsize(20)
        axes[r, c].yaxis.offsetText.set_fontsize(20)
plt.show()
```
