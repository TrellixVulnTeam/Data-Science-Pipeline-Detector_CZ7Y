### Figuring out how the filter() function works with lambda functions

```python
# adapted from https://www.geeksforgeeks.org/lambda-filter-python-examples/
# Python Program to find numbers divisible by 3 from a list using anonymous function
  
# Take a list of numbers. 
my_list = [12, 65, 54, 39, 102, 339, 221, 50, 70, ]
  
# use anonymous function to filter and comparing 
# if divisible or not
result = list(filter(lambda x: x % 3 == 0, my_list)) 
  
print(result) 

# explaining the above:
# the filter function takes a function and iterable as arguments: filter(function, iterable)
# 'function' gets run for every item in 'iterable'
# our lambda function returns True every time a number is divisible by 3
# every item for which the lambda function returns True, gets returned by 'filter'
```

### Trying the above with a dictionary

```python
mydict={'num1':5,'num2':4,'num3':3,'num4':2,'num5':1}
search_key='num1'
selected_dict=dict(filter(lambda dict_item: search_key in dict_item,mydict.items()))
print(selected_dict)
print(selected_dict.values())
```

### Plotting some EEG data

```python
# Code in this cell was partially taken from:
# https://matplotlib.org/3.3.1/gallery/specialty_plots/mri_with_eeg.html

# I (lizelle niit) in turn modified this cell from https://www.kaggle.com/code/ahernandez1/american-epilepsy-society-seizure-a-playground

from matplotlib.collections import LineCollection

#file_to_inspect = scipy.io.loadmat('/kaggle/input/seizure-prediction/Patient_1/Patient_1/Patient_1_interictal_segment_0034.mat')
file_to_inspect = scipy.io.loadmat('/kaggle/input/seizure-prediction/Dog_4/Dog_4/Dog_4_interictal_segment_0481.mat')
# retrieves only the item which contain the data of interest
search_key = '_segment_'
segment = dict(filter(lambda item: search_key in item[0], file_to_inspect.items()))
segment = list(segment.values())
data = segment[0][0][0][0]
print('the dimensions of the data matrix', data.shape)
num_electrodes = data.shape[0]             # rows (i.e. electrodes) in the data matrix
n_samples = data.shape[1]                  # number of samples on each row (i.e. electrode's samples)
electrode_names = segment[0][0][0][3][0]   # name or labels of the electrodes
print(electrode_names)
t = 10 * np.arange(n_samples) / n_samples # we know that n_samples comprise 10 minutes of measurements. so we can scale the numbers accordingly.

myfig, myaxs = plt.subplots(num_electrodes,figsize=(100,100))
myfig.suptitle('Vertically stacked subplots')
for plot_i in range(num_electrodes):
    myaxs[plot_i].plot(t,data[plot_i,:])


plt.show()
```
