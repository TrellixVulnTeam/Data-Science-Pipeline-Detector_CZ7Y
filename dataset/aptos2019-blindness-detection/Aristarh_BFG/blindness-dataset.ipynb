```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
import cv2
```

```python
PATH = '../input/aptos2019-blindness-detection/'
PATH_IMG = '../input/aptos2019-blindness-detection/train_images/'
```

```python
df = pd.read_csv(f'{PATH}train.csv')
df.head()
```

```python
df.shape
```

```python
df.diagnosis.hist()
```

```python
def get_images(diagnosis):
    fig, ax = plt.subplots(1,3,figsize=(15, 15))
    for i in range(3):
        id = random.choice(df.loc[df['diagnosis'] == diagnosis].id_code.reset_index(drop=True))
        img = cv2.imread(f'{PATH_IMG}{id}.png')[...,[2,1,0]]
        ax[i].set_title(f'Диагноз {str(diagnosis)}')
        ax[i].imshow(img)
    plt.show()
```

```python
get_images(0)
get_images(1)
get_images(2)
get_images(3)
```
