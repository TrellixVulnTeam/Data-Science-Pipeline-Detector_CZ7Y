## Image size distribution

```python
target_dirs = ['../input/train_images', '../input/test_images']
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(20,20), sharex=True)
for target_dir, ax in zip(target_dirs, [ax1, ax2]):
    # imagesize library doesn't read whole image, but metadata.
    shapes = [[str(imagesize.get("{}/{}".format(target_dir, path)))] for path in os.listdir(target_dir)]

    shapes_df = pd.DataFrame.from_records(shapes, columns=['image_size'])
    shapes_df['image_size'] = pd.Categorical(shapes_df.image_size)
    sns.countplot(shapes_df['image_size'], ax=ax).set_title(target_dir) 
    
```

## Can you as human classify them?

```python
def display_by_diagnosis(diagnosis):
    target_class_df = train_df[train_df['diagnosis'] == diagnosis]

    images = [Image(filename='../input/train_images/{}.png'.format(target_class_df.iloc[i].id_code), width=100, height=100) for i in range(0, 5)]
    display(*images)    
```

### No DR

```python
display_by_diagnosis(diagnosis=0)
```

### Proliferative DR

```python
display_by_diagnosis(diagnosis=4)
```
