```python
import numpy as np
import pandas as pd

lb_men = pd.read_csv("../input/march-machine-learning-mania-leaderboards/mens-march-mania-2022-privateleaderboard.csv")
lb_women = pd.read_csv("../input/march-machine-learning-mania-leaderboards/womens-march-mania-2022-privateleaderboard.csv")

count_dict = {}
score_dict = {}

# Probably not the cleanest way but it gets the job done :D
for lb in [lb_men, lb_women]:
    for i, row in lb.iterrows():

        if not row.TeamName in count_dict:
            count_dict[row.TeamName] = 0
            score_dict[row.TeamName] = 0
        
        count_dict[row.TeamName] += 1
        score_dict[row.TeamName] += row.Score

final_dict = {}

for k in count_dict.keys():
    if count_dict[k] == 2:
        final_dict[k] = score_dict[k] / 2

final_sorted = np.array(sorted(final_dict.items(), key=lambda x: x[1]))
final_sorted_pd = pd.DataFrame({'Ranking' : np.arange(len(final_sorted)), 'Team' : final_sorted[:, 0], 'Score' : final_sorted[:, 1].astype(np.float64)}).set_index('Ranking')
final_sorted_pd.to_csv('merged_private_leaderboard.csv')

with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(final_sorted_pd)
```
