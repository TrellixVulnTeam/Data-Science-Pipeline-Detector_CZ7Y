```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import os

from collections import defaultdict

from IPython.display import Markdown, display


pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 80)
```

```python
def printmd(string):
    display(Markdown(string))
```

### List all datasets

```python
os.listdir('../input/mens-march-mania-2022/MDataFiles_Stage1/')
```

* To avoid reading each data frame separately, I have used dictionary approach.
* This is a round about of pandas limitation. In pandas, we can not assign data frame names dynamically.
* In the dictionary approach, keys name are the name of data frames and values are the data frames.

```python
data_dir                   = '../input/mens-march-mania-2022/MDataFiles_Stage1/'
df_dict                    = defaultdict()

for aFile in os.listdir(data_dir):
    file_name                = aFile.split('.')[0]
    if file_name == 'MTeamSpellings':
        file_encoding = 'cp1252'
    else:
        file_encoding = 'utf-8'
    df_dict[file_name]       = pd.read_csv(data_dir + aFile, encoding = file_encoding)
    print('{} data shape : {}'.format(file_name, df_dict[file_name].shape))
```

### **Look at few examples of each data**

```python
for aKey in df_dict.keys():
    printmd('### Samples of Data : {}'.format(aKey))
    print(df_dict[aKey].sample(5))
```
