```python
import pandas as pd
tournament_results_df = pd.read_csv('../input/mens-march-mania-2022/MDataFiles_Stage1/MNCAATourneyCompactResults.csv')
```

```python
tournament_results2016_df = tournament_results_df.query("Season >= 2016")
tournament_results2016_df
```

Check again, sample submission format is as follows, consisting of ID and Pred.

```python
sample_submission.head()
```

```python
for key, row in tournament_results2016_df.iterrows():
    if row['WTeamID'] < row['LTeamID']:
        # Check season_win_lost type
        id_name = str(row['Season']) + '_' + str(row['WTeamID']) + '_' + str(row['LTeamID'])
        sample_submission.loc[sample_submission['ID'] == id_name, 'Pred'] = 0.89 #replace with 1.0
    else:
        # Check season_lost_win type
        id_name = str(row['Season']) + '_' + str(row['LTeamID']) + '_' + str(row['WTeamID'])
        sample_submission.loc[sample_submission['ID'] == id_name, 'Pred'] = 0.11 #replace with 0.0
```
