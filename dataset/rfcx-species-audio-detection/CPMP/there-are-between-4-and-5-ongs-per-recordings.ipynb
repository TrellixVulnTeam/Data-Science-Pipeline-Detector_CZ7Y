# There are between 4 and 5 songs per recording on average.

The sample submission get a score of 0.185.  What can we get out of this?

Frist, let's import what we need. The competition metric is said to be similar to sklean `label_ranking_average_precision_score` except for rows weights.  IF we have a single row then they should be equal.

```python
import numpy as np
from sklearn.metrics import label_ranking_average_precision_score
```

let's compute what a constant 0.5 submission, as in sample submission, would yield with rows of varying number of songs.

```python
y_pred = 0.5 * np.ones((1, 24))
for i in range(25):
    y_true = np.zeros((1, 24))
    y_true[0, :i] = 1
    score = label_ranking_average_precision_score(y_true, y_pred)
    print(i, '%0.3f' % score, y_true)
```

We see that sample submisison score of 0.185 lies between the scores with 4 or 5 targets equal to one.

Wait a minute.  Is the ordering of targets relevant?  No, according to the metric definition.  

Let's check by reversing the order of targets.

```python
y_pred = 0.5 * np.ones((1, 24))
for i in range(25):
    y_true = np.zeros((1, 24))
    y_true[0, i:] = 1
    score = label_ranking_average_precision_score(y_true, y_pred)
    print(i, '%0.3f' % score, y_true)
```

We see that 0.185 still lies between the scores of row with 4 and 5 targets equal to one.
