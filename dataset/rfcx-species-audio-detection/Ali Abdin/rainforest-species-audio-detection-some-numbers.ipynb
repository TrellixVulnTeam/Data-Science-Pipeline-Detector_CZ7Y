Check entry types in train_tp and train_fp

```python
# Check if there are more than 1 entries in audiofile found
count_multiple = 0
# Check if there are more than 1 species in the audiofile detected
count_different_spec = 0
for recording_id in train_tp["recording_id"]:
    if len(train_tp[train_tp["recording_id"] == recording_id]) > 1:
        count_multiple = count_multiple + 1
    if len(set(train_tp[train_tp["recording_id"] == recording_id]["species_id"])) > 1:
        count_different_spec = count_different_spec + 1
```

```python
(count_multiple/len(train_tp),count_different_spec/len(train_tp), count_different_spec/count_multiple, 1-(count_multiple/len(train_tp)))
```

How many recordings are in train_tp and train_fp

```python
recordings = []
for recording_id in train_tp["recording_id"]:
    if len(train_fp[train_fp["recording_id"] == recording_id]) > 0:
        recordings.append(recording_id)
```

How many samples are completly correctly recognized and therefore not in train_fp

```python
len(train_tp["recording_id"].unique()) - len(set(recordings))
```

```python
train_tp_complete = list(set(list(train_tp["recording_id"])) - set(recordings))
```

Create DataFrame with only true positives which are only in train_tp and not in train_fp

```python
train_tp_comp = pd.DataFrame()
for rec_id in train_tp_complete:
    chunk = train_tp[train_tp["recording_id"] == rec_id]
    train_tp_comp = pd.concat([train_tp_comp, chunk])
```

```python
train_tp_comp
```

```python
# Check if there are more than 1 entries in audiofile found
count_multiple = 0
# Check if there are more than 1 species in the audiofile detected
count_different_spec = 0
for recording_id in train_tp_comp["recording_id"]:
    if len(train_tp_comp[train_tp_comp["recording_id"] == recording_id]) > 1:
        count_multiple = count_multiple + 1
    if len(set(train_tp_comp[train_fp["recording_id"] == recording_id]["species_id"])) > 1:
        count_different_spec = count_different_spec + 1
```

```python
(count_multiple/len(train_fp),count_different_spec/len(train_fp), count_different_spec/count_multiple, 1-(count_multiple/len(train_fp)))
```

Almost all of them have only 1 entry and none of them have multiple species
