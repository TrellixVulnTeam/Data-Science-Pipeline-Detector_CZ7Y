```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 
import pandas as pd
import soundfile as sf
%matplotlib inline
plt.rcParams['figure.figsize'] = (12, 5);
sns.set_style('whitegrid')
```

```python
train_fp=pd.read_csv("../input/rfcx-species-audio-detection/train_fp.csv")
train_tp=pd.read_csv("../input/rfcx-species-audio-detection/train_tp.csv")
train_tp.species_id.unique()
```

```python
train_tp.shape, train_fp.shape
train_tp.recording_id.nunique(), train_fp.recording_id.nunique()
```

```python
# true_pos is showing if species are labeled correctly or not. 
train_fp["true_pos"]=0
train_tp["true_pos"]=1
train=pd.concat([train_fp, train_tp], ignore_index=True)
```

```python
#no of common audio files
len(set(train_tp["recording_id"]) & set(train_fp["recording_id"]))
```

```python
train.head()
```

```python
#All audio have same number of frames i.e. 2880000.
train_tp['nframes'] = train_tp['recording_id'].apply(lambda f: len(sf.read("../input/rfcx-species-audio-detection/train/"+f+".flac")[0]))
train_tp.head()
```

```python
category_group = train[['recording_id','species_id', 'true_pos']].groupby(['species_id', 'true_pos']).count()
plot = category_group.unstack().reindex(category_group.unstack().sum(axis=1).sort_values().index).plot(kind='bar', stacked=True, title="Number of Audio Samples per Category", figsize=(12,5))
plot.set_xlabel("Category")
plot.set_ylabel("Number of Samples");
```

```python
# reading first 100 frames of audio
data, rate = sf.read("../input/rfcx-species-audio-detection/train/00204008d.flac",frames=100)
```

```python
# Specgram plot of first 100 frames.
# Specgram computes the windowed discrete-time Fourier transform of a signal using a sliding window.
Pxx, freqs, bins, im = plt.specgram(data, Fs=rate)

# add axis labels
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
```
