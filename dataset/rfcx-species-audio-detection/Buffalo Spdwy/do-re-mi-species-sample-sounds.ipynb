```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Audio,display
import librosa
```

### it still feels amazing to me, the variety of sounds we hear can be each represented by a single time series

```python
sr    = 48000 # sample rate

# do re mi fa sol la ti do, my MUSIC 117 works!
do    = 261.6
seq   = np.exp([np.log(do) + np.log(2)/12 * x for x in np.cumsum([0,2,2,1,2,2,2,1])])
wav   = 4e-1 * np.concatenate([np.sin( 2 * np.pi/sr * s * np.arange(sr)) for s in seq])

plt.figure(figsize=(15,5))
plt.plot(wav,'.-')                              
plt.axvline(sr,color='r',ls='-')
plt.xlim(int(sr*0.99),int(sr*1.01))
plt.title('wave plot at the transition between "do" and "re"')
plt.xlabel('sample')
plt.ylabel('pressure')
plt.grid(True)

display(Audio( wav ,rate=sr,normalize=False))
```

### By a pretrained human brain, many species' characteristics can be learnt by less than 20 clips.

```python
train_tp = pd.read_csv('/kaggle/input/rfcx-species-audio-detection/train_fp.csv')
train_tp.head()
```

```python
for sid,df in train_tp.groupby('species_id'):
    print('='*10 + f'species_id {sid}' + '='*10)
    agg = []
    for index,row in df.sample(20).iterrows():
        wav, sr = librosa.load(f'/kaggle/input/rfcx-species-audio-detection/train/{row["recording_id"]}.flac', sr=None)
        clip    = wav[int(row["t_min"]*sr):int(row["t_max"]*sr)]
        agg.append(clip / np.max(np.abs(clip)) * 2)        
        agg.append(np.sin( 2 * np.pi/sr * 8 * do * np.arange(int(sr/10)))) # beep
    display(Audio(np.concatenate(agg),rate=sr))
```
