```python
train_tp_df = pd.read_csv("../input/rfcx-species-audio-detection/train_tp.csv")
train_fp_df = pd.read_csv("../input/rfcx-species-audio-detection/train_fp.csv")
```

```python
sub = pd.read_csv("../input/rfcx-species-audio-detection/sample_submission.csv")
```

```python
temp_counts_tp = train_tp_df.species_id.value_counts()
temp_counts_fp = train_fp_df.species_id.value_counts()
```

```python
for col in sub.columns[1:]:    
    sub[col] =  0.9* (temp_counts_tp[temp_counts_tp.index == int(col[1:])].values[0] / train_tp_df.shape[0]) + \
    0.1*(temp_counts_fp[temp_counts_fp.index == int(col[1:])].values[0] / train_fp_df.shape[0])
```

```python
sub.to_csv("sample_submission.csv", index=False)
```
