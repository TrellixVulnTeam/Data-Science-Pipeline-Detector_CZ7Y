## This notebooks shows how to extract MFCC feature
* Extract features first, reduce pre-process data time.
* We can use MFCC feature into CNN model.
* Dataset will public for competitors use.
* Public MFCC dataset https://www.kaggle.com/super13579/rcsadmfcc/

### import packages

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)


import os
import librosa
import soundfile as sf
```

### Load training data

```python
train_f = pd.read_csv("../input/rfcx-species-audio-detection/train_fp.csv")
train_t = pd.read_csv("../input/rfcx-species-audio-detection/train_tp.csv")
submit = pd.read_csv("../input/rfcx-species-audio-detection/sample_submission.csv")
```

```python
train_df = pd.concat([train_f,train_t]).reset_index(drop=True)
```

### MFCC Extract function
* we can change fs and mfcc dimention

```python
def MFCC_Extract(path,fs=551, mfcc_dim=20):
    x , sr = sf.read(path)
    mfccs = librosa.feature.mfcc(x, n_mfcc=mfcc_dim, sr=fs)
    return mfccs
```

### Data save as npy file

```python
from tqdm import tqdm
train_MFCC = []
os.makedirs("./train_mfcc")
for i in tqdm(train_df['recording_id']):
    audio_path = f"../input/rfcx-species-audio-detection/train/{i}.flac"
    mfcc = MFCC_Extract(audio_path)
    np.save(f"./train_mfcc/{i}.npy",mfcc)
    del mfcc
```
