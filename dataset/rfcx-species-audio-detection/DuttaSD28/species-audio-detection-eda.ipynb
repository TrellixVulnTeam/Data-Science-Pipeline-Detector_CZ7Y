# Species Audio Detection -Aim
The dataset consists of data for a lot of species in a forest. We are to predict the species present in a particular audio file. In this kernel, we will be analysing the data and trying to create an initial model.

Sources:
* [Kernel by Bojan Tunguz](https://www.kaggle.com/tunguz/rainforest-rapids-baseline)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import librosa
```

```python
# input training csv file audio labels(True Positives)
truePositiveData = pd.read_csv("../input/rfcx-species-audio-detection/train_tp.csv")
truePositiveData.head()
```

# EDA

```python
#Check for null values
truePositiveData.isnull().any()
```

```python
# Are some records repeated or not?
truePositiveData['recording_id'].nunique() == len(truePositiveData)
```

Possible causes for lengths not matching:
* There are duplicates
* There are multiple species for some audio. Each species for one audio has been put separately.

```python
# Unique species id Count Plot
plt.figure(figsize = (10, 5), dpi = 300)
sns.countplot(truePositiveData['species_id'], palette='autumn')
plt.title('species counts');
```

**Inference**

The Distribution of species if mostly uniform. However, species 23 has a high number of occurences, so we need to probe further.

```python
# Song type analysis
plt.figure(figsize = (10, 5), dpi = 300)
sns.countplot(truePositiveData['songtype_id'], palette='gist_earth')
plt.title('songtype_id Counts');
```

**Inference**

There is a huge difference between the songtype id's. Class 1 has about 1000+ counts while class 4 has about 150 or so.
This might create a problem for models.

```python
# Sense of the numerical data
truePositiveData.describe()
```

```python
# Distribution of f_min and f_max
plt.figure(figsize = (10, 5), dpi = 300)
plt.style.use('ggplot')
sns.distplot(truePositiveData['f_min'], color='black')
sns.distplot(truePositiveData['f_max'], color='red')
plt.title('Min and Max frequencies')
plt.legend(['Min_Frequency', 'Max_Frequency']);
```

```python
# log of f_min
plt.figure(figsize = (10, 5), dpi = 300)
plt.style.use('seaborn-paper')
max_of_min_freq = truePositiveData['f_min'].max()
sns.distplot(np.exp(truePositiveData['f_min'] / max_of_min_freq), color='olive')
plt.title('exp(truePositiveData["f_min"] / max_of_min_freq)');
```
