```python
#%pip install librosa
import librosa
import soundfile as sf
import scipy.signal as signal
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data_dir = '/kaggle/input/rfcx-species-audio-detection/'
df1=pd.read_csv(data_dir+'train_tp.csv')
show = df1.iloc[0]
df1.head()
```

Read the FLAC file associated with the first row of df1. Look at the signal from t_min to t_max.

```python
data, samplerate = sf.read(data_dir+'train/'+show.recording_id+'.flac')
times = np.linspace(0,len(data),len(data))/samplerate
plt.plot(times,data)
plt.xlim(show.t_min, show.t_max)
plt.ylim(-.1,.1)
plt.ylabel('Time (s)')
plt.show()
```

Look at the spectrogram of the data, using the limits of t and f.

```python
Pxx, freqs, bins, im = plt.specgram(data, Fs=samplerate)

# add axis labels
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.xlim(44,46)
plt.ylim(2000,6000)
plt.show()
```

We can also play the file and listen for ourselves!

```python
import IPython.display as ipd
ipd.Audio(data,rate=samplerate)
```

An alternative approach to plotting spectrograph, based on https://www.kdnuggets.com/2020/02/audio-data-analysis-deep-learning-python-part-1.html

```python
from librosa import display as ld
X = librosa.stft(data)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14, 5))
ld.specshow(Xdb, sr=samplerate, x_axis='time', y_axis='hz')
plt.colorbar()
plt.show()
```
