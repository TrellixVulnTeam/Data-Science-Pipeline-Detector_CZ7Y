```python
import os
os.listdir('../input/jigsaw-unintended-bias-in-toxicity-classification')

PATH = '../input/jigsaw-unintended-bias-in-toxicity-classification'
```

```python
import pandas as pd
```

```python
train = pd.read_csv(PATH + '/train.csv')
```

```python
from tqdm.auto import tqdm

tqdm.pandas()
```

```python
a = " ".join(train.comment_text.values)
```

```python
tt_list = a.split(' ')
```

```python
train_text_list = train['comment_text'].progress_map(lambda x: x.split())
```

```python
from itertools import chain

word_list = list(chain.from_iterable(train_text_list))
```

```python
d = {}
for word in tqdm(word_list):
    if d.get(word) is None:
        d[word] = 1
    else:
        d[word] += 1
```

```python
sorted_values = sorted(d.items(), key=lambda x: x[1], reverse=True)
```
