```python
sorted_df = df.sort_values(by=['severe_toxicity', 'toxicity'], ascending=False)
sorted_df.head()
```

```python
print(sorted_df.count())
```

```python
filtered_df = sorted_df[(sorted_df['severe_toxicity'] > 0.8) & (sorted_df['toxicity'] > 0.8)]
filtered_df[['comment_text', 'severe_toxicity', 'toxicity']]
```

```python
severe_toxic_df = sorted_df[(sorted_df['severe_toxicity'] > 0.3)]
print(severe_toxic_df['id'].count())
severe_toxic_df[['comment_text', 'severe_toxicity', 'toxicity']]
```

```python
toxic_df = sorted_df[(sorted_df['toxicity'] > 0.9)]
print(toxic_df['id'].count())
# toxic_df[['comment_text', 'severe_toxicity', 'toxicity']]
toxic_df.head(20)[general + culture_labels]
```

```python
filtered_df = sorted_df[(sorted_df['comment_text'].str.len() <= 140)]
print(len(filtered_df))
filtered_df.head(20)
```

```python
toxic_df = filtered_df[(filtered_df['toxicity'] >= 0.8)]
print(len(toxic_df))
toxic_df.head(40)[general + culture_labels]
```

```python
middle_toxic_df = filtered_df[(filtered_df['toxicity'] < 0.55) & (filtered_df['toxicity'] > 0.45)]
print(len(middle_toxic_df))
middle_toxic_df.head(40)[general + culture_labels]
```

```python
low_toxic_df = filtered_df[(filtered_df['toxicity'] < 0.2)]
print(len(low_toxic_df))
low_toxic_df.head(40)[general + culture_labels]
```

```python
lower_toxic_df = filtered_df[(filtered_df['toxicity'] < 0.1)]
print(len(lower_toxic_df))
lower_toxic_df.head(40)[general + culture_labels]
```
