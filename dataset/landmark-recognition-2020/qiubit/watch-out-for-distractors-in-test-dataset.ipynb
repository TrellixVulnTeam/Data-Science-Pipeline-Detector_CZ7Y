But maybe Gaia Germani is also a landmark ;)

Actually, there is a way to check that. If you click on photos of Gaia Germani and Saint Nicolas Tower on Wikipedia, you can find out that they correspond to the following categories:

Gaia Germani
- http://commons.wikimedia.org/wiki/Category:I_complessi
- http://commons.wikimedia.org/wiki/Category:Films_by_Luigi_Filippo_D%27Amico
- http://commons.wikimedia.org/wiki/Category:1965_films
- http://commons.wikimedia.org/wiki/Category:Gaia_Germani

Saint Nicolas Tower:
- http://commons.wikimedia.org/wiki/Category:Tour_Saint-Nicolas

Google published mapping from train landmark ids to those categories. It's available at https://s3.amazonaws.com/google-landmark/metadata/train_label_to_category.csv

Let's download it and load into pandas to see whether we can find correct landmark ids.

```python
!wget https://s3.amazonaws.com/google-landmark/metadata/train_label_to_category.csv
```

```python
TRAIN_LABEL_TO_CATEGORY = 'train_label_to_category.csv'

GAIA_CATEGORIES = [
    'http://commons.wikimedia.org/wiki/Category:I_complessi',
    'http://commons.wikimedia.org/wiki/Category:Films_by_Luigi_Filippo_D%27Amico',
    'http://commons.wikimedia.org/wiki/Category:1965_films',
    'http://commons.wikimedia.org/wiki/Category:Gaia_Germani'
]

TOWER_CATEGORIES = [
    'http://commons.wikimedia.org/wiki/Category:Tour_Saint-Nicolas'
]
```

```python
import pandas as pd
```

```python
df = pd.read_csv(TRAIN_LABEL_TO_CATEGORY)
```

```python
landmark_categories = set(df.category.values)
```

Now, that all category names are loaded, let's check whether Gaia Germani or Saint Nicolas Tower is indeed a landmark.

```python
gaia_is_landmark = False

for cat_name in GAIA_CATEGORIES:
    if cat_name in landmark_categories:
        gaia_is_landmark = True

if gaia_is_landmark:
    print('Gaia Germani IS a landmark')
else:
    print('Gaia Germani IS NOT a landmark')
```

```python
tower_is_landmark = False

for cat_name in TOWER_CATEGORIES:
    if cat_name in landmark_categories:
        tower_is_landmark = True

if tower_is_landmark:
    print('Saint Nicolas Tower IS a landmark')
else:
    print('Saint Nicolas Tower IS NOT a landmark')
```

This experiment shows that, with high confidence, distractors are indeed present in the test dataset. You should avoid making any predictions for such distractors, since it will lower your Leaderboard score.
