```python
# Freeze layers from 0 to start and unfreeze next
def set_trainable(model, start):
    layers_number = len(model.layers)
    for i in range(0,start):
        model.layers[i].trainable = False
    for i in range(start, layers_number):
        model.layers[i].trainable = True

# Create callback to save best model
def save_best_callback(path, monitor="val_loss", mode="min"): 
    return tf.keras.callbacks.ModelCheckpoint(filepath=path, save_weights_only=True, monitor=monitor, mode=mode, save_best_only=True)

# Plot history of .fit
def plot_history(history, metrics=None):
    labels = []
    df = pd.DataFrame(history)
    ncols = len(df.columns)
    fig, axs = plt.subplots(ncols=ncols // 2, nrows=1, figsize=(35,10))
    if metrics is None:
        metrics = set([i.replace("val_", "") for i in history.keys()])
    for metric in metrics:
        labels.append([metric, f"val_{metric}"])
    if ncols == 2:
        sns.lineplot(data=df, ax=axs)
        plt.xlabel("epoch")
    else:
        for pair_ind in range(len(labels)):
            sns.lineplot(data=df[labels[pair_ind]], ax=axs[pair_ind % ncols])
            axs[pair_ind % ncols].set_xlabel("epoch")
    plt.show()

def plot_wordcloud(text, figsize=(15,15), **kwargs):
    _, ax = plt.subplots(figsize=(15,15))
    wordcloud = WordCloud(**kwargs).generate(text)
    ax.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off");

def expand_contractions(text):
    return " ".join([ct_dict[w] if w in ct_dict else w for w in text.split()])
```

```python
html_pattern = re.compile(r"<[^<]+?>")
date_pattern = re.compile(r"\d\d[-\.\\\/]\d\d[-\.\\\/]\d{0,4}")
emoticon_pattern = re.compile((f"({'|'.join(EMOTICONS.keys())})"))
url_pattern = re.compile(r'https?://\S+|www\.\S+') 
chat_words_map_dict = {pair[0]:pair[1] for pair in [line.split("=") for line in chat_words_str.lower().split("\n")]}
chat_words_set = set(chat_words_map_dict.keys())

def remove_html(text):
    return html_pattern.sub("", text)

def remove_dates(text):
    return date_pattern.sub("", text)

def remove_punctuation(text):
    return "".join(filter(lambda x: x not in string.punctuation, text))

def remove_nums(text):
    return "".join(filter(lambda x: not x.isdigit(), text))

def replace_emoticons(text):
    return emoticon_pattern.sub("", text)

def chat_words_conversion(text):
    words_to_replace = set(text.split()).intersection(chat_words_set)
    return " ".join(chat_words_map_dict.get(word, word) for word in text.split())
                               
def replace_emoji_with_word(text):
    return emoji.demojize(text)

def clear_text(text):
    return chat_words_conversion(remove_punctuation(replace_emoji_with_word(replace_emoticons(remove_nums(remove_dates(remove_html(text)))))))
```

```python
def build_model_new():
    checkpoint = "bert-base-multilingual-uncased"
    loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
    metrics = [tf.keras.metrics.AUC(name="AUC")]
    learning_rate = tf.keras.optimizers.schedules.ExponentialDecay(1e-5, 1000, 0.98)
    optimizer = Adam(learning_rate=learning_rate)
    
    input_ids_in = tf.keras.layers.Input(shape=(MAX_LENGTH,), name='input_ids', dtype='int32')
    input_masks_in = tf.keras.layers.Input(shape=(MAX_LENGTH,), name='attention_mask', dtype='int32') 
    input_token_types_in = tf.keras.layers.Input(shape=(MAX_LENGTH,), name='token_type_ids', dtype='int32') 
    
    bert_model = TFAutoModelForSequenceClassification.from_pretrained(checkpoint, num_labels=1)
    
    embedding_layer = bert_model(input_ids_in, token_type_ids=input_token_types_in, attention_mask=input_masks_in)[0]
    cls_token = embedding_layer[:, 0, :]
    
    X = tf.keras.layers.Dense(50, activation='relu')(cls_token)
    X = tf.keras.layers.Dropout(0.2)(X)
    X = tf.keras.layers.Dense(4, activation='sigmoid')(X)
    
    model = tf.keras.Model(inputs=[input_ids_in, input_masks_in, input_token_types_in], outputs = X)
    model.compile(optimizer, loss=loss, metrics=metrics)
    
    return model

def build_model():
    checkpoint = "bert-base-multilingual-uncased"
    loss = tf.keras.losses.BinaryCrossentropy(from_logits=True)
    metrics = [tf.keras.metrics.AUC(name="AUC")]
    learning_rate = tf.keras.optimizers.schedules.ExponentialDecay(1e-5, 1000, 0.98)
    optimizer = Adam(learning_rate=learning_rate)
    
    model = TFAutoModelForSequenceClassification.from_pretrained(checkpoint, num_labels=1)
    model.compile(optimizer, loss=loss, metrics=metrics)
    
    return model
```
