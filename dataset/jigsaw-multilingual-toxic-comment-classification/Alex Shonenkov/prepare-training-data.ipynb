```python
import numpy as np
import pandas as pd 
```

```python
%%time

df_train = pd.read_csv(f'../input/jigsaw-multilingual-toxic-comment-classification/jigsaw-unintended-bias-train.csv')[['id', 'comment_text', 'toxic']]
df_train = df_train[(df_train['toxic'] < 0.2) | (df_train['toxic'] > 0.5)]
df_train = df_train.append(
    pd.read_csv(f'../input/jigsaw-multilingual-toxic-comment-classification/jigsaw-toxic-comment-train.csv')[['id', 'comment_text', 'toxic']]
)
df_train['lang'] = 'en'

for lang in ['es', 'tr', 'it', 'ru', 'pt', 'fr']:
    df_lang = pd.read_csv(f'../input/jigsaw-train-multilingual-coments-google-api/jigsaw-toxic-comment-train-google-{lang}-cleaned.csv')[[
        'id', 'comment_text', 'toxic', 
    ]]
    df_lang['lang'] = lang
    df_train = df_train.append(df_lang)


df_train = df_train[~df_train['comment_text'].isna()]
df_train = df_train.drop_duplicates(subset='comment_text')
df_train['toxic'] = df_train['toxic'].round().astype(np.int32)
```
