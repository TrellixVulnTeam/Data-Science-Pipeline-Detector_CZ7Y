```python
import time
import os
import psutil
import pandas as pd
import numpy as np
from tqdm import tqdm

def count_time(func):
    def int_time():
        start_time = time.time()
        func()
        over_time = time.time()
        total_time = over_time - start_time
        tps = COUNT/total_time
        print("The Tps: %s Item/s" % tps)
    return int_time
```
