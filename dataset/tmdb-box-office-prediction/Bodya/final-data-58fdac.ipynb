```python
def get_dict(x):
    try: 
        d = eval(x)
    except:
        d = {}
    return d

def extract(col):
    train[col] = train[col].map(lambda x: get_dict(x))
    train[col] = train[col].map(lambda x: sorted([y['name'] for y in x]))\
        .map(lambda x: ','.join(map(str, x)))
    
    test[col] = test[col].map(lambda x: get_dict(x))
    test[col] = test[col].map(lambda x: sorted([y['name'] for y in x]))\
         .map(lambda x: ','.join(map(str, x)))

def extract_count(col):
    extract(col)
    train[col+'_count'] = train[col].str.split(pat = ',')\
            .str.len().replace(np.nan, 0).astype(int)
          
    test[col+'_count'] = test[col].str.split(pat = ',')\
            .str.len().replace(np.nan, 0).astype(int)
    
    
def length(col):
    train[col+'_length'] = train[col].str.len().replace(np.nan, 0).astype(int)
    test[col+'_length'] = test[col].str.len().replace(np.nan, 0).astype(int)
    train[col+'_word_count'] = train[col].str.split().str.len().replace(np.nan, 0).astype(int)
    test[col+'_word_count'] = test[col].str.split().str.len().replace(np.nan, 0).astype(int)
```

```python
extract('belongs_to_collection')

extract_count('genres')
extract_count('spoken_languages')
extract_count('production_countries')
extract_count('production_companies')
extract_count('Keywords')
extract_count('cast')

train['crew'] = train['crew'].map(lambda x: get_dict(x))
test['crew'] = test['crew'].map(lambda x: get_dict(x))

train['crew_all_names'] = train['crew'].map(lambda x: sorted([y['name'] for y in x]))\
        .map(lambda x: ','.join(map(str, x)))
    
test['crew_all_names'] = test['crew'].map(lambda x: sorted([y['name'] for y in x]))\
         .map(lambda x: ','.join(map(str, x)))


train['crew_count'] = train['crew_all_names'].str.split(pat = ',')\
            .str.len().replace(np.nan, 0).astype(int)   
test['crew_count'] = test['crew_all_names'].str.split(pat = ',')\
            .str.len().replace(np.nan, 0).astype(int)

length('original_title')
length('title')
length('Keywords')
length('overview')
length('tagline')
```

```python
def get_job(job, x):
    temp = []
    for i in x:
        if i['job'] == job: temp.append(i)
    return temp

def get_dep(dep, x):
    temp = []
    for i in x:
        if i['department'] == dep: temp.append(i)
    return len(temp)
```

```python
train['producers'] = train['crew'].map(lambda x: get_job('Producer', x))
test['producers'] = test['crew'].map(lambda x: get_job('Producer', x))

train['producers'] = train['producers'].map(lambda x: sorted([y['name'] for y in x]))\
        .map(lambda x: ','.join(map(str, x)))
test['producers'] = test['producers'].map(lambda x: sorted([y['name'] for y in x]))\
        .map(lambda x: ','.join(map(str, x)))
```
