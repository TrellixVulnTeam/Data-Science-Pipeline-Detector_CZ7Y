```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pandas_profiling
import seaborn as sns
```

```python
#importing the TMBD Training Dataset
train = pd.read_csv('../input/tmdb-box-office-prediction/train.csv')
```

```python
#Showing training Data
train.head()
```

```python
#importing test dataset 
test = pd.read_csv('../input/tmdb-box-office-prediction/test.csv')
```

```python
#Showing Test Data
test.head()
```

```python
train.info()
```

```python
train.describe()
```

```python

X = train.iloc[:, np.r_[2:3,9:10,14]].values
y = train.iloc[:, -1].values
```

```python
X_test = test.iloc[:, np.r_[2:3,9:10,14]].values
```

```python
# Taking Care of Missing data
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean')
imputer.fit(X[:,np.r_[2:3]])
X[:,np.r_[2:3]] = imputer.transform(X[:,np.r_[2:3]])
```

```python
#Taking care of missing data in Test Data set
imputer1 = SimpleImputer(missing_values = np.nan, strategy = 'mean')
imputer1.fit(X_test[:, np.r_[2:3]])
X_test[:, np.r_[2:3]] = imputer1.transform(X_test[:, np.r_[2:3]])
```

```python
X[:, np.r_[2:3]].shape
```

```python
# Performing Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X[:, 0:] = sc.fit_transform(X[:, 0:])
X_test[:, 0:]=sc.transform(X_test[:, 0:])
```

```python

from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 125, random_state = 100)
rf_model = regressor.fit(X,y)
print("rf: ", rf_model.score(X, y))
```

```python
y_pred = rf_model.predict(X_test)
sub = pd.DataFrame({'id': test.id, 'revenue': y_pred})
sub.to_csv('sub_rf.csv', index=False)
```
