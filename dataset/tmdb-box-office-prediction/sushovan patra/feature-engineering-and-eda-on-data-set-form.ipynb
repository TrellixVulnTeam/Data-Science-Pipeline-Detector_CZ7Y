```python
def get_dict(s):
    try:
        d = eval(s)
    except:
        d = {}
    return d
```

#### SPOKEN LANGUAGES

English spoken movies has more chances of revenue as well as storing no of spoken language as a factor

```python
language=[]
number=[]
for i in train['spoken_languages']:
    d=get_dict(i)
    l=list(pd.DataFrame(d)['name'])
    language.append(l)
    number.append(len(l))

train['spoken_languages']=language

train['no_of_spoken_languages']=number
```

```python
k=[]
for i in train['spoken_languages']:
    if 'English' not in i:
        k.append(0)
    else:
        k.append(1)
```

###### Now USA concidered 1, UK concidered 2, if no usa or uk among list of countries concidered 0,if both usa and uk then3

```python
k=[]
for i in train['production_countries']:
    if ('United Kingdom'  in i) and ('United States of America'  in i):
        k.append(3)
    elif('United States of America'  in i):
        k.append(1)
    elif ('United Kingdom'  in i):
        k.append(3)
    else:
        k.append(0)
```

```python
train['production_countries']=k
```
