```python
train_df = pd.read_csv("/kaggle/input/titanic/train.csv")
train_df.head()
```

```python
test_df = pd.read_csv("/kaggle/input/titanic/test.csv")
test_df.head()
```

```python
test_df.shape
```

```python
train_df.shape
```

```python
df1 = pd.concat([train_df,test_df])
   
```

```python
df=df1.drop(["Survived"],axis=1)
```

```python
df.head(50)
```

```python
y_train = train_df["Survived"].values

```

```python
last_df = df.drop(["Cabin", "Name", "Ticket"], axis=1)
```

```python
last_df
```

```python
last_df["Embarked"].value_counts()
```

```python
last_df["Embarked"].fillna("S", inplace=True)

```

```python
last_df["Age"].fillna(last_df["Age"].mean(), inplace=True)
```

```python
last_df.isna().sum()
```

```python
last_df["Fare"].fillna(last_df["Fare"].mean(), inplace=True)
```

```python
last_df.isna().sum()
```

```python
last_df["Embarked"] = last_df["Embarked"].map({"S": 3, "C": 2, "Q": 1}).astype(int)
```

```python
final_df1=pd.get_dummies(last_df["Sex"])
```

```python
final_df1
```

```python
final_df2=final_df1.drop(["male"],axis=1)
```

```python
final_df3 = last_df.drop(["Sex"],axis=1)
```

```python
final_df3
```

```python
final_df=pd.concat([final_df3,final_df2],axis=1)
```

```python
final_df
```

```python
final_df.isna().sum()
```

```python
X_train = final_df[:y_train.shape[0]]
```

```python
X_train
```

```python
X_test = final_df[y_train.shape[0]:]
```

```python
import random
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, test_size = 0.25, random_state=42)
```

```python
final_df
```

```python
import xgboost as xgb
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
from xgboost import XGBClassifier
```

```python
gbm_params = {"n_estimators": [5, 10, 15, 20, 25], 
    "max_depth": [3, 5, 7, 9, 11, 13]}
```

```python
def set_seed(seed_value):
    random.seed(seed_value)
    np.random.seed(seed_value)
    os.environ["PYTHONHASHSEED"] = str(seed_value)
    

SEED = 42
set_seed(SEED)
```

```python
model_xgb = xgb.XGBClassifier(
    random_state=SEED,)
```

```python
model_xgb = GridSearchCV(
    model_xgb, 
    gbm_params, 
    cv=5,
    scoring='accuracy',)
```

```python
model_xgb.fit(X_train, y_train)
```

```python
model_xgb = GradientBoostingClassifier(n_estimators=25, 
    max_depth=11)
```

```python
model_xgb_tuned =  model_xgb.fit(X_train,y_train)
```

```python
y_pred = model_xgb_tuned.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
from xgboost import XGBClassifier
```

```python
model = XGBClassifier(subsample = 0.7, max_depth = 4,booster='gbtree')
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
```

```python
print("Accuracy:%.2f%%" % (accuracy * 100.0))
```
