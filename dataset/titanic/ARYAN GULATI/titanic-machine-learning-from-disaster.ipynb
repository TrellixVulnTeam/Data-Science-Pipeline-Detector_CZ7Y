### Cleaning Dataset

```python
data.info()
```

```python
data.columns
```

```python
columns_to_drop=[ 'Name','Ticket','Cabin', 'Embarked']
```

```python
data_clean=data.drop(columns=columns_to_drop)
```

```python
data_clean
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
data_clean['Sex']
```

```python
# Fit label encoder and return encoded labels of column Sex
le.fit_transform(data_clean['Sex'])
```

```python
data_clean['Sex']
```

```python
data_clean['Sex']=le.fit_transform(data_clean['Sex'])
```

```python
data_clean
```

```python
data_clean.info()
```

```python
data_clean["Age"].median()
```

```python
data_clean["Fare"].median()
```

```python
data_clean["Age"]=data_clean["Age"].fillna(value=data_clean["Age"].median())
```

```python
data_clean["Fare"]=data_clean["Fare"].fillna(value=data_clean["Fare"].median())
```

```python
data_clean.info()
```

```python
data_clean.columns
```

```python
input_cols=['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']
```

```python
output_cols=['Survived']
```

```python
X=data_clean[input_cols]
```

```python
y=data_clean[output_cols]
```

```python
X.columns
```

```python
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
```

```python
model = DecisionTreeClassifier(max_depth=5, )
```

```python
model.fit(X,y)
```

```python
model.score(X,y)
```
