<p style = "font-size:18px; ; text-align: center" >There comes situations in real life when we need to make some decisions and based on these decisions, we decide what should we do next. Similar situations arise in programming also where we need to make some decisions and based on these decisions we will execute the next block of code. Decision-making statements in programming languages decide the direction of the flow of program execution. <br>
In Python, if else elif statement is used for decision making. </p>

<a id="8.1"></a>
<p style = "font-size:30px; color:#331340 ; text-align: left; border-radius: 5px 5px; padding: 5px" ><strong>IF Statement</strong></p>

<p style = "font-size:18px; ; text-align: center" >If statement is the most simple decision-making statement. It is used to decide whether a certain statement or block of statements will be executed or not i.e if a certain condition is true then a block of statement is executed otherwise not. </p>

```python
# example:
num=50
if num > 200:
    print("Num is greater than 20")
print(" it didin't executed because the condituon was not true")
```

<a id="8.2"></a>
<p style = "font-size:30px; color:#331340 ; text-align: left; border-radius: 5px 5px; padding: 5px" ><strong>Else statement</strong></p>

<p style = "font-size:18px; ; text-align: center" >The if statement alone tells us that if a condition is true it will execute a block of statements and if the condition is false it won‚Äôt. But what if we want to do something else if the condition is false. Here comes the else statement. We can use the else statement with if statement to execute a block of code when the condition is false. </p>

```python
num=50
if num > 200:
    print("Num is greater than 200")
else:
    print("Num is less than 200")
```

<a id="8.3"></a>
<p style = "font-size:30px; color:#331340 ; text-align: left; border-radius: 5px 5px; padding: 5px" ><strong>IF-ELIF-ELSE</strong></p>

<p style = "font-size:18px; ; text-align: center" >Here, a user can decide among multiple options. The if statements are executed from the top down. As soon as one of the conditions controlling the if is true, the statement associated with that if is executed, and the rest of the ladder is bypassed. If none of the conditions is true, then the final else statement will be executed. </p>

```python
num=20
if num == 10:
    print(" Num is 10")
elif num ==15:
    print(" Num is 15")
elif num == 20:
    print(" Num is 20")
else:
    print(" Num is Greater than 20")
```

<p style = "font-size:18px; ; text-align: center" >A class method is a method which is bound to the class and no the object of the class.<br>
@classmethod decorator is used to create a class method</p>

```python
class Employee:
    comapny="Google"
    salary=2000
    
    def chng_salary(self,sal):
        self.salary=sal
    
      

sam=Employee()
print(sam.salary)
sam.chng_salary(100)

print(sam.salary)

Employee.salary
```

<p style = "font-size:18px; ; text-align: center" >Here as we can see instance of the class changes but the attribute of the class remains same and what if we want to change the attribute of the class. so we will use class methods</p>

```python
# 1st Method

class Employee:
    comapny="Google"
    salary=2000
    
    def chng_salary(self,sal):
        self.__class__.salary=sal # we can use .__class__. also called dender class
    
      

sam=Employee()
print(sam.salary)
sam.chng_salary(100)

print(sam.salary)

Employee.salary
```

```python
# 2nd Method

class Employee:
    comapny="Google"
    salary=2000
    
    @classmethod #we can use decorator.Decorator takes input as fxn and modifies it
    def chng_salary(cls,sal): 
        cls.salary=sal 
    
      

sam=Employee()
print(sam.salary)
sam.chng_salary(100)

print(sam.salary)

Employee.salary
```

Enumerate() method adds a counter to an iterable and returns it in a form of enumerating object. This enumerated object can then be used directly for loops or converted into a list of tuples using the list() method.</p>

<p style = "font-size:30px; color:#331340 ; text-align: left; border-radius: 5px 5px; padding: 5px" ><strong>2. List Comprehension</strong></p>

<p style = "font-size:18px; ; text-align: center" >List comprehensions are used for creating new lists from other iterables like tuples, strings, arrays, lists, etc. A list comprehension consists of brackets containing the expression, which is executed for each element along with the for loop to iterate over each element. .</p>

```python
# print numbers which are greater than 4
l1=[5,4,2,5,3,1,2,4]
[x for x in l1 if x > 4]
```

<p style = "font-size:30px; color:#331340 ; text-align: left; border-radius: 5px 5px; padding: 5px" ><strong>3. Lambda Functions</strong></p>

<p style = "font-size:18px; ; text-align: center" >Python Lambda Functions are anonymous function means that the function is without a name. As we already know that the def keyword is used to define a normal function in Python. Similarly, the lambda keyword is used to define an anonymous function in Python. </p>

```python
# function for printing cube of a number
cube = lambda x : x*x*x
cube(5)
```

```python
summ = lambda a,b,c: a+b+c
summ(2,3,4)
```

<p style = "font-size:30px; color:#331340 ; text-align: left; border-radius: 5px 5px; padding: 5px" ><strong>4. Map Functions</strong></p>

<p style = "font-size:18px; ; text-align: center" >This function accepts another function and a sequence of ‚Äòiterables‚Äô as parameters and provides output after applying the function to each iterable in the sequence.</p>

```python
# want to print cube of a list
cube = lambda x : x*x*x # first make lambda fxn
list1 = [1,2,4,5] # list we want to find cube
list(map(cube,list1))

```

```python
# increment a number
list1=[1,2,3,4,5]
list(map(lambda x : x+1,list1))
```

```python
list1=[1,2,3,4,2,8,2]
print(list(map(lambda x:x > 3,list1)))
```

<p style = "font-size:30px; color:#331340 ; text-align: left; border-radius: 5px 5px; padding: 5px" ><strong>5. Filter Functions</strong></p>

<p style = "font-size:18px; ; text-align: center" >The filter() function is used to generate an output list of values that return true when the function is called.</p>

```python
list1=[1,2,3,4,2,8,2]
print(list(filter(lambda x:x > 3,list1)))
```
