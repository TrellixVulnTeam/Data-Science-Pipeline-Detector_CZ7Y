# Pandas

```python
import pandas as pd
```

```python
#creating a series
n=[1,2,3,4,5]
ind=(1,2,3,4,5)
s1=pd.Series(n,index=ind)
s1
```

```python
b=[12,23,1,34,3,5,22]
s2=pd.Series(b)
s2
```

```python
# operations on series
s4=s1.append(s2)
s4
```

```python
s1.add(s2)
```

```python
s5=s1.sub(s2)
s5.div(s2)
```

```python
s5.max()
```

```python
s2.min()
```

```python
s2.median()
```

```python
s2.mean()
```

```python
s5.mean()
```
