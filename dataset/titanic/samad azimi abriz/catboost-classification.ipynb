# Missing Values

```python
train.isnull().sum()
```

```python
100*(train.isnull().sum()/len(train))
```

```python
def missing_percent(train):
    nan_percent= 100*(train.isnull().sum()/len(train))
    nan_percent= nan_percent[nan_percent>0].sort_values()
    return nan_percent
```

```python
nan_percent= missing_percent(train)
```

```python
nan_percent
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=nan_percent.index, y=nan_percent)
plt.xticks(rotation=90)
```

```python
train.drop("Cabin", axis = 1, inplace = True)
```

# Split Data into train,test

```python
X = train.drop("Survived", axis = 1, inplace = False)
y = train["Survived"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

# Normalization Data¶

```python
scaler=StandardScaler()
scaler.fit(X_train)
```

```python
scaler_train=scaler.transform(X_train)
scaler_test=scaler.transform(X_test)
```

```python
from catboost import CatBoostClassifier as cbc
```

# catboost classification

```python
model=cbc()
model.fit(scaler_train,y_train)
```

```python
y_pred=model.predict(scaler_test)
```

```python
pd.DataFrame({'Y test':y_test,'Y predict':y_pred})
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
