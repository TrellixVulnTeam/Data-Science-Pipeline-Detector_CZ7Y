# Building a Machine Learning Model

```python
train.head()
```

```python
sex=pd.get_dummies(train["Sex"],drop_first=True)
embarked=pd.get_dummies(train["Embarked"],drop_first=True)
new_data=pd.concat([train,sex,embarked],axis=1)
new_data.head()
```

```python
# Drop the some columns
new_data.drop(columns=["Name","Sex","Ticket","Embarked"],axis=1,inplace=True)
```

```python
new_data.head()
```

```python
#Defined X value and y value , and split the data train

X = new_data.drop(['Survived'],axis=1)
y = new_data.Survived

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=40)
```

```python
print("X Train : ", X_train.shape)
print("X Test  : ", X_test.shape)
print("Y Train : ", y_train.shape)
print("Y Test  : ", y_test.shape)
```

```python
# using the model Linear Regression

LinearRegression_model=LinearRegression(fit_intercept=True,normalize=False,copy_X=True, n_jobs=None)

# fit model

LinearRegression_model.fit(X_train,y_train)

# Score X and Y - test and train

print("Score the X-train with Y-train is : ", LinearRegression_model.score(X_train,y_train))
print("Score the X-test  with Y-test  is : ", LinearRegression_model.score(X_test,y_test))

# Expected value Y using X test
y_pred_LR=LinearRegression_model.predict(X_test)

# Model Evaluation
print( " Model Evaluation Linear R : mean absolute error is ", mean_absolute_error(y_test,y_pred_LR))
print(" Model Evaluation Linear R : mean squared  error is " , mean_squared_error(y_test,y_pred_LR))
print(" Model Evaluation Linear R : median absolute error is " ,median_absolute_error(y_test,y_pred_LR)) 

print(" The Score Linear Regression " ,r2_score(y_test,y_pred_LR)) 
```

```python
# using the model Logistic Regression

LogisticRegression_model=LogisticRegression(solver='liblinear')

# fit model

LogisticRegression_model.fit(X_train,y_train)

# Score X and Y - test and train

print("Score the X-train with Y-train is : ", LogisticRegression_model.score(X_train,y_train))
print("Score the X-test  with Y-test  is : ", LogisticRegression_model.score(X_test,y_test))

# Expected value Y using X test
y_pred_LO=LogisticRegression_model.predict(X_test)

# Model Evaluation
print( " Model Evaluation Linear R : mean absolute error is ", mean_absolute_error(y_test,y_pred_LO))
print(" Model Evaluation Linear R : mean squared  error is " , mean_squared_error(y_test,y_pred_LO))
print(" Model Evaluation Linear R : median absolute error is " ,median_absolute_error(y_test,y_pred_LO)) 

```

```python
# using the Decision Tree Classifier

DecisionTreeClassifier_model=DecisionTreeClassifier()

# fit model

DecisionTreeClassifier_model.fit(X_train,y_train)

# Score X and Y - test and train

print("Score the X-train with Y-train is : ", DecisionTreeClassifier_model.score(X_train,y_train))
print("Score the X-test  with Y-test  is : ", DecisionTreeClassifier_model.score(X_test,y_test))

# Expected value Y using X test
y_predDTC=DecisionTreeClassifier_model.predict(X_test)

# Model Evaluation

print(" The Score Decision Tree Regressor model " ,r2_score(y_test,y_predDTC)) 
```

```python
# using the Random Random Forest Classifier

RandomForestClassifier_model=RandomForestClassifier(n_estimators=10000,
    criterion='gini',
    max_depth=None,
    min_samples_split=2,
    min_samples_leaf=1,
    min_weight_fraction_leaf=0.0,
    max_features='auto',
    max_leaf_nodes=None,
    min_impurity_decrease=0.0,
    min_impurity_split=None,
    bootstrap=True,
    oob_score=False,
    n_jobs=None,
    random_state=None,
    verbose=0,
    warm_start=False,
    class_weight=None,
    ccp_alpha=0.0,
    max_samples=None)

# fit model

RandomForestClassifier_model.fit(X_train,y_train)

# Score X and Y - test and train

print("Score the X-train with Y-train is : ", RandomForestClassifier_model.score(X_train,y_train))
print("Score the X-test  with Y-test  is : ", RandomForestClassifier_model.score(X_test,y_test))

# Expected value Y using X test
y_predRFR=RandomForestClassifier_model.predict(X_test)

# Model Evaluation

print(" The Score Random Forest Regressor Model " ,r2_score(y_test,y_predRFR)) 
```

```python
# using the XGB Classifier Model

XGBClassifier_model=XGBClassifier()

# fit model

XGBClassifier_model.fit(X_train,y_train)

# Score X and Y - test and train

print("Score the X-train with Y-train is : ", XGBClassifier_model.score(X_train,y_train))
print("Score the X-test  with Y-test  is : ", XGBClassifier_model.score(X_test,y_test))

# Expected value Y using X test
y_predXGB=XGBClassifier_model.predict(X_test)

# Model Evaluation

print(" The Score MLP Regressor Model " ,r2_score(y_test,y_predXGB)) 
```
