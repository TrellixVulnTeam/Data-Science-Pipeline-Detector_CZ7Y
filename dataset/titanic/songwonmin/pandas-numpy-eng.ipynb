````python
test["Fare"] = test["Fare"].fillna(test["Fare"].mean())
train["Embarked"] = train["Embarked"].fillna(train["Embarked"].mode()[0])
````

````python
test["Fare"].isnull().sum()
````

**평균이나 최빈값을 이용하거나 아예 새로운 값으로 지정하여 결측치를 표현하기도 합니다.**

**이 코드에서는 가격의 결측값을 가격의 평균으로 대체하고 승선위치의 결측값을 가장 많은 사람이 승선한 위치로 대체하였습니다.**

**Express missing values by using the average or mode value or by designating them as completely new values.**
