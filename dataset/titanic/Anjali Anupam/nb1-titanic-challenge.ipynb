#  Predicting survival of passengers using Decision Tree

### Importing all the necessary libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
from pandas import DataFrame
```

### Reading the given datasets

```python
titanic_train = pd.read_csv('../input/titanic/train.csv') 
titanic_test = pd.read_csv('../input/titanic/test.csv')
test_ids = titanic_test['PassengerId']
```

```python
titanic_train.head()
```

### Filling all the NaN values

```python
titanic_train = titanic_train.drop(['Name','Ticket','Fare','Cabin'],axis='columns')
cols = ['SibSp','Parch','Age']
for col in cols: 
    titanic_train[col].fillna(titanic_train[col].median(),inplace=True)
    
titanic_train.Embarked.fillna('U',inplace=True)
titanic_train.head()
```

```python
titanic_test = titanic_test.drop(['Name','Ticket','Fare','Cabin'],axis='columns')
cols = ['SibSp','Parch','Age']
for col in cols: 
    titanic_test[col].fillna(titanic_test[col].median(),inplace=True)
    
titanic_test.Embarked.fillna('U',inplace=True)
titanic_test.head()
```

### Using LabelEncoder to convert strings into numbers

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_sex = LabelEncoder()
le_embarked = LabelEncoder()
```

```python
titanic_train['Sex']= le_sex.fit_transform(titanic_train['Sex'])
titanic_train['Embarked']= le_embarked.fit_transform(titanic_train['Embarked'])
```

```python
titanic_train.head()
```

```python
titanic_test['Sex']= le_sex.fit_transform(titanic_test['Sex'])
titanic_test['Embarked']= le_embarked.fit_transform(titanic_test['Embarked'])
```

```python
y_train = titanic_train['Survived']
x_train = titanic_train.drop('Survived',axis='columns')
x_train
```

### Predicting the survival using Decision Tree

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier(criterion='entropy',random_state=0)
```

```python
model.fit(x_train,y_train)
```

```python
submission_pred = model.predict(titanic_test)
```

### Converting results into csv file

```python
df = DataFrame(test_ids,columns=['PassengerId'])
df['Survived'] = submission_pred
df.info
```

```python
df.to_csv('submission.csv',index=False)
```
