> # Cat Plots

Catplot can handle 8 different plots currently available in Seaborn. catplot function can do all these types of plots and one can specify the type of plot one needs with the kind parameter.

So it is kind of one stop for every plot you will require for bivariate analysis.
(Univariate analysis plots like displots are not included)

The default kind in catplot() is “strip”, corresponding to stripplot(). Here are the list of different type of plots, involving categorical variables, you can make with catplot and the names of the kind.

1. Categorical scatterplots with catplot
    * stripplot() – with kind=”strip”
    * swarmplot() – with kind=”swarm”
1. Categorical distribution plots with catplot
    * boxplot() – with kind=”box”
    * violinplot() – with kind=”violin”
    * boxenplot() – with kind=”boxen”
1. Categorical estimate plots with catplot
    * pointplot() – with kind=”point”
    * barplot() – with kind=”bar”
    * countplot() – with kind=”count”


Basic parmeters:

sns.catplot(data= df, 
               x= 'promo',
               y= 'sales',
               kind= 'fill with your plot type')

Lets take some examples. Here I am goung to use same old school titanic dataset for a few plots.

```python
import pandas as pd
import seaborn as sns
train = pd.read_csv("../input/titanic/train.csv")
test = pd.read_csv("../input/titanic/test.csv")
```

Lets start with the bar plot between "SibSp" and "Survived"
* kind = "bar"

```python
sns.catplot(x="SibSp",y="Survived",data=train,kind="bar",height=4)
```

Moving to the point plot

```python
sns.catplot(x="SibSp",y="Survived",data=train,kind="point",height=4)
```

Lets create some swarm plots between "Survived" and "Fare"

```python
sns.catplot(x="Survived",y="Fare",data=train,kind="swarm",height=4)
```

Similarly other plots can be easily created for bivariate analysis.
* Official documentation: https://seaborn.pydata.org/generated/seaborn.catplot.html

## Changing Labels
Update: I had a  little trouble while setting x and y labels in catplot this can be easily done by using the below syntax:

```python
g =sns.catplot(x="Survived",y="Fare",data=train,kind="swarm",height=4)
g.set_axis_labels("Survived (Categorical)","Fare (Numerical)")
```
