```python
# Column with null values and their count
c = 0
is_null = []
for i in df.columns:
    if df[i].isnull().sum()>0:
        is_null.append(i)
        print(i,df[i].isnull().sum())
        c = c+1
print('Number of columns containing null values are:',c)
```

```python
x = df.drop(['Sex','Embarked'],axis = 1)
def count_outliers(data,col):
        q1 = data[col].quantile(0.25,interpolation='nearest')
        q2 = data[col].quantile(0.5,interpolation='nearest')
        q3 = data[col].quantile(0.75,interpolation='nearest')
        q4 = data[col].quantile(1,interpolation='nearest')
        IQR = q3 -q1
        global LLP
        global ULP
        LLP = q1 - 1.5*IQR
        ULP = q3 + 1.5*IQR
        if data[col].min() > LLP and data[col].max() < ULP:
            print("No outliers in",i)
        else:
            print("There are outliers in",i)
            x = data[data[col]<LLP][col].size
            y = data[data[col]>ULP][col].size
            a.append(i)
            print('Count of outliers are:',x+y)
global a
a = []
for i in x.columns:
    count_outliers(df,i)
```

## Replacing null values with mean median or mode

```python
# If your null value column contain outliers then fill them with medain or else mean
# If your null value column is categorical then fill them with mode
for i in is_null:
    if df[i].dtype == 'int64' or df[i].dtype == 'float64':
        if i in a:
                df[i].fillna(df[i].median(),inplace=True)
        else:
            df[i].fillna(df[i].mean(),inplace=True)
    elif df[i].dtype == 'object':
        df[i].fillna(df[i].mode()[0],inplace=True)
```
