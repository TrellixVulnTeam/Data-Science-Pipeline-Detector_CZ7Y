<h1 style = "font-family: Snell Roundhand;font-weight: 500;font-size: 30px; font-style: normal; background-color: #FFFB30; color :black;"> 1) What is NumPy ?</h1>

* Numpy stands for numerical python
* NumPy is a Python library used for working with arrays.
* It also has functions for working in domain of linear algebra, fourier transform, and matrices.
* NumPy was created in 2005 by Travis Oliphant. It is an open source project and you can use it freely.
* NumPy stands for Numerical Python.


<h1 style = "font-family: Snell Roundhand;font-weight: 500;font-size: 30px; font-style: normal; background-color: #FFFB30; color :black;"> 2) Why Numpy?</h1>
Numpy arrays offer the following benefits over Python lists for operating on numerical data:

* Numpy is effictive way to handal the wast amount of data 
* In Python we have lists that serve the purpose of arrays, but they are slow to process.
* NumPy aims to provide an array object that is up to 50x faster than traditional Python lists.
* The array object in NumPy is called ndarray, it provides a lot of supporting functions that make working with ndarray very easy.
* Arrays are very frequently used in data science, where speed and resources are very important.

<h1 style = "font-family: Snell Roundhand;font-weight: 500;font-size: 30px; font-style: normal; background-color: #FFFB30; color :black;"> 7) Attributes</h1>

**Now we take a example to print Diminsion and shape of a given array**
* In array there is **ndim** function to find Diminsion of array

* we called **shape** to find shape of a given array

* There is a **reshape** fun to reshape the array

* **size** return total number of the array, this is equal to product of shape

* **dtype** descrive the type of the elements in array

```python
ar=np.array([[10,12,14],[1,2,3]])
print("Array :\n",ar)
print("Diminsion of array : ",ar.ndim)
print("shape of a given array : ",ar.shape)
print("reshaping the array :")
ar1=ar.reshape(3,2)
print("Reshaped array : \n",ar1)
print("Size of the array : ",ar.size)
print("type of the elements in array : ",ar.dtype)
```

<h1 style = "font-family: Snell Roundhand;font-weight: 500;font-size: 30px; font-style: normal; background-color: #FFFB30; color :black;"> 8) Find the max, min and sum of the array</h1>

*Numpy provides the **max()** , **min()** and **sum()** function which are used to find maximum, minimum,and sum of the array element respectaivilly*

```python
arr=np.array([[1,2,3],[6,5,7],[9,8,0]])
print("3d Array :\n",arr)
print("maximum element : ",arr.max())
print("minimum element : ",arr.min())
print("sum of elements : ",arr.sum())

    
```

<h1 style = "font-family: Snell Roundhand;font-weight: 500;font-size: 30px; font-style: normal; background-color: #FFFB30; color :black;"> 9) Find square root and standerd devision</h1>

* **sqrt** find the squre root of the array element

* **std** use to find standerd devision of the array element


```python
print("3d Array :\n",arr)
print("squre root of array elements : ",np.sqrt(arr))
print("Stenderd devision : ",np.std(arr))
```

<h1 style = "font-family: Snell Roundhand;font-weight: 500;font-size: 30px; font-style: normal; background-color: #FFFB30; color :black;"> 10) Arithmetic operation on the array</h1>

*The numpy modules allow as to perform the arithmetic operation on multi dimensional array directly*
* You can see in the following example

```python
ar=np.array([[11,12,32],[10,20,30],[15,25,35]])
ar1=np.array([[2,4,6],[3,5,7],[9,8,7]])
print("1st array :- \n" ,ar)
print("2ed array :- \n",ar1)
print("sum of two arrays :-\n",ar+ar1)
print("subtract of two arrays :-\n",ar-ar1)
print("multiplecation of two arrays :-\n",ar*ar1)
print("Product of two arrays :-\n",ar-ar1)
```

<h1 style = "font-family: Snell Roundhand;font-weight: 500;font-size: 30px; font-style: normal; background-color: #FFFB30; color :black;"> 11) Array cancatination</h1>

*numpy provides us with the vertical stacking and horizontal stacking which allow us to cancatination of multi dimensional array*
* **np.vstack()** use for vertical cancatination
* **np.hstack()** use for horizontal cancatination

```python
print("Vertical cancatination :\n",np.vstack((ar,ar1)))
print("horizontal cancatination :\n",np.hstack((ar,ar1)))
```

<h1 style = "font-family: Snell Roundhand;font-weight: 500;font-size: 30px; font-style: normal; background-color: #FFFB30; color :black;"> 12) Numpy dtype</h1>

*All the items of a numpy array are data type objects also known as numpy dtypes.*
*A data type object implements the fixed size of memory corresponding to an array*


**Syntex:-**

numpy.dtype(column_header,data-type,copy)

**Example:-**

```python
d=np.dtype([("Salary",np.float),("da",np.int)])
arr=np.array([(10000.12,4500),(20000.15,5600)],dtype=d)
print("Salary : ",arr["Salary"])
print("da : ",arr["da"])
print("Sum of da : \n",sum(arr["da"]))
print("max of salary : \n",max(arr["Salary"]))
```

<h1 style = "font-family: Snell Roundhand;font-weight: 500;font-size: 30px; font-style: normal; background-color: #FFFB30; color :black;"> 13) NumPy array Creation</h1>

*The ndarray object can be constructed by using the following routines*

* **numpy.empty :-** it will return a empty array
* **numpy.zeros :-** it will return array with zeros
* **numpy.ones :-**  it will return array with ones
### all these syntex accepts following parameters:-
* **shape** desird shape of the specified array
* **dtype** data type of the array items. The difault is the float
* **order** specified order of a array


```python
print("printing empty array :")
arr=np.empty((3,3),dtype=int)
print(arr)
print("printing Array with zeros :")
ar=np.zeros((3,3),dtype=int)
print(ar)
print("printing Array with ones :")
a=np.ones((3,3),dtype=int)
print(a)
```

<h1 style = "font-family: Snell Roundhand;font-weight: 500;font-size: 30px; font-style: normal; background-color: #FFFB30; color :black;"> 14) numpy array with existing data</h1>

*numpy provides us the way to create an array by using the existing data*

**syntax:-**
*numpy.asarray(sequence,dtype=None,order=None)*

**Examples:-**

```python
print("creating numpy array using the list :")
list=[1,2,3,4,5,6,7]
array=np.asarray(list)
print(array)
print("data type : ",type(array))
print("**************************")
print("creating numpy array using tuple : ")
tup=(7,6,5,4,3,2,1)
arr=np.asarray(tup)
print(arr)
print("data type : ",type(arr))
```
