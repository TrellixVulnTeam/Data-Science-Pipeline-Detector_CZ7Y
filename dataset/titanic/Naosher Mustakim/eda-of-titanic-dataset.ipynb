# <span style="color:#0096FF;"> Let's look at the dataset: </span>

```python
data = pd.read_csv('/kaggle/input/titanic/train.csv')
data.head()
```

```python
new_columns=['Survived', 'Pclass', 'Name', 'Sex', 'Age', 'SibSp',
       'Parch', 'Ticket', 'Fare', 'Cabin', 'Embarked']
```

<span style="color:#0096FF;"> Let's look at the correlation between data. </span>

```python
new_data=data[new_columns]
new_data.corr()
```

```python
f ,ax = plt.subplots(figsize = (12, 10))
colormap = plt.cm.YlOrRd_r
plt.title('Pearson Correlation of Features', size=20, weight="bold")
res = sns.heatmap(new_data.corr(), linewidths=0.1, vmax=1.0,cmap=colormap,square=True, annot= True, annot_kws={"size": 16}, ax=ax)
res.set_yticklabels(res.get_ymajorticklabels(), fontsize = 18, rotation=5, weight="bold")
res.set_xticklabels(res.get_xmajorticklabels(), fontsize = 18, weight="bold")
plt.show()
```

## <span style="color:#0096FF;">  Lets see total passenger count in each class. </span>

```python
data[['Pclass', 'Survived']].groupby(['Pclass'], as_index=False).count().sort_values(by='Pclass', ascending=True)
```

```python
plt.figure(figsize=(10,6))
font= {'family':'serif','color':'darkred','size':35}
pl = sns.color_palette("gist_rainbow", 9) 
# making error in palette style name gives a list of styles 
ax= sns.countplot(x=data['Pclass'], data=data, palette=pl)
ax.bar_label(ax.containers[0])
plt.xlabel("Pclass", fontdict = font)
plt.ylabel("Count", fontdict = font)
plt.show();
```

<span style="color:#0096FF;"> So class 3 had the highest number of passengers and class 2 had the lowest. </span>

## <span style="color:#0096FF;"> Deceased vs Survivor count.</span>

```python
data['Survived'].groupby(data['Pclass']).value_counts()
```

```python
plt.figure(figsize=(10,6))
font= {'family':'serif','color':'darkred','size':35}
pl = sns.color_palette("gist_rainbow", 9)
fig = sns.countplot(x=data['Pclass'], hue = data['Survived'], data=data, palette=pl)
fig.bar_label(fig.containers[0])
fig.bar_label(fig.containers[1])
plt.xlabel("Pclass", fontdict = font)
plt.ylabel("Count", fontdict = font)
plt.show();
```

<span style="color:#0096FF;"> Above figure clearly shows that ratio of death among passengers is higher for class 3. Then class 2 had the next highest death ratio. Class 1 had the lowest death ratio. Infect more people survived than died in class 1. Clearly, socio-economic class played a vital role in one's survival in the titanic. </span> 
* <span style="color:#0096FF;"> Now question is:</span> <span style="color:#EE4901;"> What is the percentage of survival for each class? </span>


## <span style="color:#0096FF;"> Percentage of survival for each class.</span>

```python
data[['Pclass', 'Survived']].groupby(['Pclass'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```

<span style="color:#0096FF;"> Almost 3 of every 4 people (75%) died in class 3. More than half died (53%) in class 2. In class 1,almost 3 of every 5 people (62%) survived.</span>

## <span style="color:#0096FF;"> Passenger distribution by gender in each class.</span>

```python
# data.Pclass.plot(kind = 'hist',bins = 50,figsize = (12,12))
plt.figure(figsize=(10,6))
font= {'family':'serif','color':'darkred','size':35}
pl = sns.color_palette("gist_rainbow", 9)
ax = sns.countplot(data =data, x = "Sex", hue ="Pclass",palette=pl)
ax.bar_label(ax.containers[0],fmt= '%0.0f')
ax.bar_label(ax.containers[1],fmt= '%0.0f')
ax.bar_label(ax.containers[2],fmt= '%0.0f')
plt.xlabel("Pclass", fontdict = font)
plt.ylabel("Count", fontdict = font)
plt.show();
```

<span style="color:#0096FF;"> Above figure shows male and female passenger counts in each class. In all class, male count is dominant.</span>

```python
plt.figure(figsize=(10,6))
sns.pointplot(x="Pclass",y="Survived", hue="Sex",data=data, palette={"male":"g", "female":"m"}, markers=["^","o"], linestyles=["-","--"])
plt.xlabel("Pclass", fontdict = font)
plt.ylabel("Survived", fontdict = font)
plt.show();
```

## <span style="color:#0096FF;">Age Distribution</span>

```python
plt.figure(figsize=(25,13))
font= {'family':'serif','color':'black','size':35}
ax = sns.distplot(a=data['Age'],kde=False);
# sns.countplot(data['Age'])
ax.bar_label(ax.containers[0],fmt= '%0.0f');
plt.xlabel("Age", fontdict = font)
plt.ylabel("Count", fontdict = font)

x = np.random.randint(low=0, high=50, size= 80)
plt.xticks(np.arange(0, len(x)+1, 5))
plt.xticks()
plt.show();
```

<span style="color:#0096FF;"> Above figure shows counts in each age group. For example, age group '0-10' had 40 passengers on board.</span>

## <span style="color:#0096FF;">Ticket farewise age distribution.</span>

```python
plt.figure(figsize=(25,13))
font= {'family':'serif','color':'darkred','size':35}
ax = sns.barplot(x=data['Age'],y = data['Fare'], hue=data['Survived'],dodge=True,ci=None)
ax.bar_label(ax.containers[0],fmt= '%0.0f')
ax.bar_label(ax.containers[1],fmt= '%0.0f')
plt.xlabel("Age", fontdict = font)
plt.ylabel("Ticket Fare", fontdict = font)
plt.xticks(rotation = 45)
plt.show();
```

## <span style="color:#0096FF;">Ticket fare distributed by survival class.</span>

```python
plt.figure(figsize=(25,12))
font= {'family':'serif','color':'darkred','size':35}
font_new= {'family':'serif','color':'red','size':25}
sns.stripplot(x="Survived", y="Fare", data=data)
plt.xlabel("Survived", fontdict = font)
plt.ylabel("Ticket Fare", fontdict = font)
plt.xticks(np.arange(2),["Deceased", "Survivor"], fontdict = font_new)
plt.yticks(size=15)
plt.show();
```
