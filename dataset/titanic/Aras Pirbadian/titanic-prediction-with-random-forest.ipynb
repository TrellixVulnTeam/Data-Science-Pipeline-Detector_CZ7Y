```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
import warnings
warnings.filterwarnings("ignore")
```

```python
train_df = pd.read_csv("/kaggle/input/titanic/train.csv")
test_df = pd.read_csv("/kaggle/input/titanic/test.csv")
df_described = train_df.describe()
```

Cleaning up the train data by dropping 'Name','PassengerId', 'Ticket' and converting 'Sex' and 'Embark' columns to numeric data
the train secuence is then scaled and 'Survived' coulmn is seperated to y_train

```python
train_df['Sex'].replace('male', 0, inplace = True)
train_df['Sex'].replace('female', 1, inplace = True)
train_df['Embark'] = pd.factorize(train_df['Embarked'])[0] + 1
train_df['Cabin'] = pd.factorize(train_df['Cabin'])[0] + 1
train_df['Age'][train_df['Age'].isnull()] = round(np.mean(train_df['Age']), 2)
train_df.drop(['Name','PassengerId', 'Ticket','Embarked'], axis = 1 , inplace = True)
train_df.isnull().sum()
y_train = train_df['Survived']
train_df.drop('Survived', axis = 1 , inplace = True)
x_train = train_df


scaler_train= StandardScaler()
scaler_train.fit(x_train)
```

Cleaning up the test data by dropping 'Name','PassengerId', 'Ticket' and converting 'Sex' and 'Embark' columns to numeric data
the test secuence is then scaled 

```python
test_df['Sex'].replace('male', 0, inplace = True)
test_df['Sex'].replace('female', 1, inplace = True)
test_df['Embark'] = pd.factorize(test_df['Embarked'])[0] + 1
test_df['Cabin'] = pd.factorize(test_df['Cabin'])[0] + 1
test_df['Age'][test_df['Age'].isnull()] = round(np.mean(test_df['Age']), 2)
test_df.drop(['Name','PassengerId', 'Ticket','Embarked'], axis = 1 , inplace = True)
#test_df.drop('Embarked', axis = 1 , inplace = True)
test_df['Fare'][test_df['Fare'].isnull()] = round(np.mean(test_df['Fare']), 2)
test_df.isnull().sum()

x_test = test_df

scaler_test= StandardScaler()
scaler_test.fit(x_test)
```

Scale all features and train with 100 trees in a random forest 

```python
x_train_scaled = scaler_train.transform(x_train)
x_test_scaled = scaler_test.transform(x_test)

Randomforest_model = RandomForestClassifier(n_estimators=100, max_features='auto', random_state=101)
Randomforest_model.fit(x_train_scaled,y_train)
```

```python
y_pred = Randomforest_model.predict(x_test_scaled)
```

Put the X-test and predictions next to each other to read them better 

```python
test_samples = x_test
test_samples['y_pred'] = y_pred.tolist()
```
