**Preprocessing**

```python
y =x["Survived"] 
```

```python
x = x.drop(['Survived'], axis=1)
```

```python
le = LabelEncoder()
x.Sex = le.fit_transform(x.Sex)
x.Name = le.fit_transform(x.Name)
x.Embarked = le.fit_transform(x.Embarked)
x.Cabin = le.fit_transform(x.Cabin)
x.Ticket = le.fit_transform(x.Ticket)
```

```python
fit_feat = SelectKBest(score_func=f_classif)
```

```python
fit_feat.fit(x,y)
```

```python
score = pd.DataFrame(fit_feat.scores_, columns=['Score Values'])
```

```python
score
```

```python
x
```

```python
x = x.drop(['PassengerId','Name','Age','SibSp','Parch'], axis=1)
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, train_size=.66, random_state=17)
```

```python
sc = StandardScaler()
```

```python
xtrain = sc.fit_transform(xtrain)
```

```python
xtest = sc.fit_transform(xtest)
```

```python
xgb = XGBClassifier()
xgb.fit(xtrain,ytrain)
```

**Modeling**

```python
rfc = RandomForestClassifier(n_estimators=400)
```

```python
rfc.fit(xtrain,ytrain)
```

```python
ypred = rfc.predict(xtest)
```

```python
accuracy_score(ypred, ytest)
```

```python
sb.heatmap(confusion_matrix(ypred,ytest),annot=True, cmap='binary')
```

```python
print(classification_report(ypred,ytest))
```

```python
plt.bar(ypred, ytest, color='y')
plt.xlabel("Predicted Value")
plt.ylabel("Tested Value")
plt.title("Accuracy Line")
```
