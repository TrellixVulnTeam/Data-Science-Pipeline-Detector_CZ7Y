    <h1><right>What is "dropna()" ?</right></h1>
</div>


```python
data.dropna()
```

dropna(): after detecting the null values as shown above, and for dealing with them, dropna() will be applied for removing NA values

```python
data.dropna(how='any')
```

```python
data.dropna(how='all')
```

how{'any', 'all'}, default 'any':
what is the differnce between them?

they determines if row or column is removed from DataFrame, when we have at least one NA or all NA.

'any' : If there is any Null values in the data frame row or column, drop that row or column.

'all' : If all values are Nulls in the dataframe column or row, then  drop that row or column.

```python
data.dropna(how='all',axis=1)
```

```python
data.dropna(how='any',axis='columns')
```

axis{0 or 'index', 1 or 'columns'}, default 0
Determine if rows or columns which contain missing values are removed.

0, or 'index : Drop rows which contain missing values.

1, or 'columns' : Drop columns which contain missing value.

AS shown above we have Null value as folow:

177 null values in "Age" column

Dealing with Age NULLS:-

```python
sns.histplot(titanic['Age'])
```

```python
int(titanic['Age'].mean())
```

the mean or the average age between all ages in the data set is 29 years old
The mean is the average of the numbers. It is easy to calculate: add up all the numbers, then divide by how many numbers there are. In other words it is the sum divided by the count.

```python
titanic['Age'].mode()
```

the mode number of Age is 24, which is the most frequent number appears
Mode: The most frequent number—that is, the number that occurs the highest number of times. Example: The mode of {4 , 2, 4, 3, 2, 2} is 2 because it occurs three times, which is more than any other number.

```python
titanic['Age'].median()
```

the median number of age is 28 years
Median: The middle number; found by ordering all data points and picking out the one in the middle (or if there are two middle numbers, taking the mean of those two numbers). Example: The median of 4, 1, and 7 is 4 because when the numbers are put in order (1 , 4, 7) , the number 4 is in the middle.

Now we can replace Null value in "Age" columns with any value we decide such as mean, mode, median or any other age.

```python
titanic["Age"].fillna(titanic["Age"].mean(), inplace = True)
```

as shown above "Age" column NULLS can be handled by filling the nan value with the mean value of the Age

```python
titanic.isnull().sum()
```

Dealing with "Cabin" Nulls

```python
titanic["Cabin"].unique()
```

we can not dealing with this type of data and replacing the null values with another "Cabin" number but we can filling it with any value such as "unknown" or "other" or any thing else

```python
titanic["Cabin"]=titanic["Cabin"].fillna("unknown")
```

for the last column that have the least number of null which is only 2 nulls, we can basicly removes it!

```python
titanic=titanic.dropna()
```
