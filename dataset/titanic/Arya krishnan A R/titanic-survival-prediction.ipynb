```python
titanic.head()
```

```python
#deleting unwanted columns
titanic.drop(['Name','PassengerId','Ticket','Cabin'], axis=1, inplace = True)
```

```python
titanic.head()
```

```python
titanic.shape #returns no of rows and columns
```

```python
#coverting into categorical data
```

```python
titanic=pd.get_dummies(titanic)
titanic.head()
```

```python
titanic['Age']=titanic['Age']/max(titanic['Age'])
titanic['Fare']=titanic['Fare']/max(titanic['Fare'])
```

```python
titanic.head()
```

```python
#checking null values
titanic.isnull().sum()
```

```python
#removing null values
titanic['Age'].fillna(titanic['Age'].median(),inplace = True)
```

```python
titanic.isnull().sum()
```

```python
titanic.head()
```

```python
titanic.shape
```

```python
#training the data
```

```python
x = titanic.drop('Survived', axis = 1)
y = titanic['Survived']
```

```python
#train-test split
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 30, random_state = 42)
```

```python
#logistic regression
```

```python
from sklearn.linear_model import LogisticRegression
lm = LogisticRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
prediction = lm.predict(x_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,prediction))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,prediction)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,prediction)
```

We have the accuracy of 83%.
