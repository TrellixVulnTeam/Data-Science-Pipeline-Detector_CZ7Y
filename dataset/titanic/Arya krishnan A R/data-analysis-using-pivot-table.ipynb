# Exploring the Titanic Dataset using Pandas in Python

```python
#importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
df = pd.read_csv('../input/titanic/train.csv')
df.head()
```

```python
#dropping few unwanted columns
df.drop(['PassengerId','Name','Ticket'],axis = 1, inplace = True)
df.head()
```

# Building a Pivot Table using Pandas

```python
#How to group data using index in a pivot table?
```

```python
#a single index
table = pd.pivot_table(data = df, index = ['Sex'])
table
```

```python
table.plot(kind ='bar')
```

```python
#multiple indexes
table = pd.pivot_table(data = df, index =['Sex','Pclass'])
table
```

```python
# Different aggregate functions
```

```python
table = pd.pivot_table(data = df, index = ['Sex','Pclass'],aggfunc={'Age':np.mean, 'Survived':np.sum})
table
```

```python
#Aggregate on specific features with values parameter
```

```python
table = pd.pivot_table(data = df,index = ['Sex','Pclass'],values = ['Survived'],aggfunc=np.mean)
table
```

value parameter is where you tell the function which features to aggregate on.

```python
table.plot(kind='bar')
```

 The survival rate of male passengers was lower than the female passengers in any given Pclass.

```python
#Find the relationship between features with columns parameter
```

```python
table = pd.pivot_table(data = df, index =['Sex'],columns =['Pclass'], values = ['Survived'],aggfunc =np.sum)
table
```

```python
table.plot(kind ='bar')
```

```python
# Handling missing data
```

```python
#display null values
table = pd.pivot_table(data = df, index =['Sex','Pclass','Survived'],columns =['Embarked'],values = ['Age'],aggfunc =np.mean)
table
```

```python
#handling null value
table = pd.pivot_table(data = df, index =['Sex','Pclass','Survived'],columns =['Embarked'],values = ['Age'],aggfunc = np.mean, fill_value =np.mean(df['Age']))
table
```
