```python
def missing_percent(df):
    nan_percent=(df.isnull().sum()/len(df))*100
    nan_percent=nan_percent[nan_percent>0].sort_values()
    return nan_percent
nan_percent=missing_percent(df)
sns.barplot(x=nan_percent.index,y=nan_percent)
```

```python
df[df['Embarked'].isnull()]
```

```python
df= df.dropna(axis=0, subset=['Embarked'])
```

```python
df=df.drop('Cabin',axis=1 )
```

```python
df.corr()['Age'].sort_values()
```

```python
df.groupby('Pclass')['Age'].mean()
```

```python
df['Age']=df.groupby('Pclass')['Age'].transform(lambda val: val.fillna(val.mean()))
```

```python
df.columns
```

```python
df_num= df[['Survived','Pclass','Age','SibSp','Parch','Fare']]
df_obj= df[['Sex','Embarked']]
```

```python
df_obj= pd.get_dummies(df_obj, drop_first=True)
```

```python
Final_df= pd.concat([df_num, df_obj], axis=1)
Final_df.shape
```

```python
Final_df.isnull()
```

```python
df=Final_df
```

```python
df.head()
```

# **Step6:Features&Label**

```python
X= df.drop('Survived', axis=1)
y= df['Survived']
```

# **Step7:Split to Train & test**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

# **Step8:Train/Test/Evaluating**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

n_estimators=[64, 100, 128, 200]
max_features=[3, 4,5,6,7,8]
bootstrap=[True, False]
param_grid= {'n_estimators':n_estimators,
            'max_features':max_features,
            'bootstrap':bootstrap}


RF_model= RandomForestClassifier()
grid=GridSearchCV(RF_model, param_grid, cv=5 )
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
y_pred= grid.predict(X_test)
```
