```python
test=pd.read_csv('../input/titanic/test.csv')
train=pd.read_csv('../input/titanic/train.csv')
```

```python
df=train.drop(['Ticket','Cabin','PassengerId'], axis=1)
df1=test.drop(['Ticket','Cabin','PassengerId'], axis=1)

print(df.head())
print(df1.head())
```

```python
title=set()
for name in train["Name"]:
    title.add(name.split(",")[1].split(".")[0].strip())
print(title)
```

```python
titles={'Sir':"officer",'Mme':"normald",'Mr':"norm",'Master':"normald",'Don':"officer",'Miss':"normald",'Lady':"normald",'Mlle':"normald",'Col':"officer",'Ms':"normald",'the Countess':"royal",'Mrs':"normald",'Major':"officer",'Capt':"officer",'Dr':"officer",'Rev':"officer",'Jonkheer':"royal"}
```

```python
df["Name"]=df["Name"].map(lambda name:name.split(",")[1].split(".")[0].strip())
df["Name"]=df.Name.map(titles)
df.head(10)
```

```python
df1["Name"]=df1["Name"].map(lambda name:name.split(",")[1].split(".")[0].strip())
df1["Name"]=df1.Name.map(titles)
df1.head(10)
```

```python
print(df.isna().sum())
print(df1.isna().sum())
```

```python
df.dropna(inplace=True)
```

```python
print(df.shape)
print(df1.shape)
```

```python
df.Sex=df.Sex.map({'female':0,'male':1})
df.Embarked=df.Embarked.map({'S':0,'Q':1,'C':2})
df.Name=df.Name.map({"royal":0,"normald":1,"officer":2,"norm":3})

df1.Sex=df1.Sex.map({'female':0,'male':1})
df1.Embarked=df1.Embarked.map({'S':0,'Q':1,'C':2})
df1.Name=df1.Name.map({"royal":0,"normald":1,"officer":2,"norm":3})

print(df.head())
print(df1.head())
```

```python
df.Age.fillna(df.Age.median(),inplace=True)
df1.Age.fillna(df1.Age.median(),inplace=True)
df1.Fare.fillna(df1.Fare.median(),inplace=True)
df1.Name.fillna(df1.Name.median(),inplace=True)
```

```python
df["Age"]=round((df.Age.mean()/df.Age.std()))
df1["Age"]=round((df1.Age.mean()/df1.Age.std()))
df1["Fare"]=round((df1.Fare.mean()/df1.Fare.std()))
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```
