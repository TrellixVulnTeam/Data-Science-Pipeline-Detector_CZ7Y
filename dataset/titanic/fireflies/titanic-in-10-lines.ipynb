```python
import h2o ; from h2o.automl import H2OAutoML ; import pandas as pd
h2o.init()

df = h2o.import_file('../input/titanic/train.csv')
df_test = h2o.import_file('../input/titanic/test.csv')
train, test = df.split_frame(ratios=[0.8], seed = 23)

x, y = train.columns, "Survived"
x.remove(y)
train[y], test[y] = train[y].asfactor(), test[y].asfactor()

aml = H2OAutoML(max_runtime_secs = 10, seed = 23)
aml.train(x = x, y = y , training_frame = train, leaderboard_frame = test)
pred = aml.leader.predict(df_test)['predict'].as_data_frame()

result = df_test['PassengerId'].as_data_frame()
result.insert(1,'Survived', pred)
result.to_csv('submission.csv', index=False)
```
