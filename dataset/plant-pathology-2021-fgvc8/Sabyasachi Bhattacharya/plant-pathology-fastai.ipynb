```python
import pandas as pd
from fastai.vision.all import *
```

```python

train_data = ImageDataLoaders.from_csv("/kaggle/input/plant-pathology-2021-fgvc8","train.csv"
                                       ,folder="train_images"
                                       ,delimiter=","
                                       ,label_delim=" "
                                       , item_tfms=Resize(224)
                                       , batch_tfms=[*aug_transforms(size=224),Normalize.from_stats(*imagenet_stats)]
                                      ,bs=32)
```

```python
train_data.show_batch()
```

```python
learn = cnn_learner(train_data,arch=models.resnet50,metrics=accuracy_multi)
```

```python
learn.fine_tune(1)
```

```python
learn.export(fname="/kaggle/working/95-acc-res-50.pkl")
```

```python
learn.show_results()
```
