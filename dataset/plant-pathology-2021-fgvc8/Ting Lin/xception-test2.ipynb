```python
import os
import numpy as np
import pandas as pd
train = pd.read_csv('../input/plant-pathology-2021-fgvc8/train.csv')
train.head()
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
mb = MultiLabelBinarizer().fit(train.labels.apply(lambda x : x.split()))
labels = pd.DataFrame(mb.transform(train.labels.apply(lambda x : x.split())), columns = mb.classes_)
new_train = pd.concat([train['image'], labels], axis=1)
new_train.head()
```

```python
from keras.preprocessing.image import ImageDataGenerator
train_datagen = ImageDataGenerator(
            rescale = 1/255.,
            rotation_range = 15,
            width_shift_range=0.2, 
            height_shift_range=0.2,
            zoom_range = 0.2,
            shear_range=0.3,
            featurewise_center=True,
            featurewise_std_normalization=True,
            horizontal_flip=True,
            vertical_flip=True,
            validation_split= 0.2,)
```

```python
seed = 72 
target_size = (384, 384)
batch_size = 8
resize_img = '../input/resized-plant2021/img_sz_384'
test_img = '../input/plant-pathology-2021-fgvc8/test_images'
submission = pd.read_csv('../input/plant-pathology-2021-fgvc8/sample_submission.csv')
submission.head()
```

```python
train_generator = train_datagen.flow_from_dataframe(
                  new_train,
                  directory = resize_img, 
                  x_col = 'image',
                  y_col = new_train.columns.tolist()[1:],
                  subset = "training",
                  color_mode = "rgb",
                  target_size = target_size,
                  class_mode = "raw",
                  batch_size = batch_size,
                  shuffle = True,
                  seed = seed,)
valid_generator = train_datagen.flow_from_dataframe(
                  new_train,
                  directory = resize_img,
                  x_col = 'image',
                  y_col = new_train.columns.tolist()[1:],
                  subset = "validation",
                  color_mode = "rgb",
                  target_size = target_size,
                  class_mode = "raw",
                  batch_size = batch_size,
                  shuffle = True,
                  seed = seed,)
```

# Test

```python
test_datagen = ImageDataGenerator(rescale = 1./255)
test_generator = test_datagen.flow_from_dataframe(
                  submission,
                  directory = test_img,
                  x_col = 'image',
                  y_col = 'labels',
                  class_mode = None,
                  batch_size=1,
                  target_size = target_size,
                  color_mode="rgb",
                  shuffle = False,
                  seed = seed,)
```

```python
import keras
model = keras.models.load_model("../input/xception4/Xception_test4.h5")
pred = model.predict(test_generator)
```

```python
perdict = (pred>0.4)
n_label = new_train.columns.tolist()[1:]
answer = []

for i in range(perdict.shape[0]):
    temp = []
    for j, k in enumerate(n_label):
        if perdict[i, j]:
            temp.append(k)
    answer.append(temp)
    
answer = [' '.join(n) for n in answer]
```

```python
submission['labels'] = np.array(answer)
submission
```

```python
submission.to_csv('submission.csv', index=False)
```
