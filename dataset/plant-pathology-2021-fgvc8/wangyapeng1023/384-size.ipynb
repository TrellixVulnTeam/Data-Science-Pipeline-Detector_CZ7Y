```python
import albumentations as A
from albumentations.core.composition import Compose, OneOf
from albumentations.augmentations.transforms import CLAHE, GaussNoise, ISONoise
from albumentations.pytorch import ToTensorV2
import matplotlib.pyplot as plt
from torchvision import transforms as T
from PIL import Image
import cv2
import os
from tqdm import tqdm
```

```python
imgResize = A.SmallestMaxSize(max_size=384,p=cv2.INTER_CUBIC)
def resize_images(path_in:str, path_out:str, filenames:list):
    if not os.path.isdir(path_out): os.mkdir(path_out)
    for filename in tqdm(filenames):
        image = cv2.imread(os.path.join(path_in, filename) )
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        image_resized = imgResize(image=image)['image']
        save_to = os.path.join(path_out, filename)
        image_resized = cv2.cvtColor(image_resized, cv2.COLOR_RGB2BGR)
        if not cv2.imwrite(filename=save_to, img=image_resized):
            print(f'Failed to save image: {filename} to dir:{path_out}')
```

```python
root_dir = '../input/plant-pathology-2021-fgvc8/train_images'
out_dir = './resize-plant-384'
resize_images(root_dir,out_dir,os.listdir(root_dir))
```
