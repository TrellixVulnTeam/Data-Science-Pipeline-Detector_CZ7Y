```python
import numpy as np
import multiprocessing as mp
from PIL import Image
import pandas as pd
from tqdm.auto import tqdm
```

```python
df = pd.read_csv("../input/plant-pathology-2021-fgvc8/train.csv")
```

```python
means = list()
stds = list()

def worker(args):
    index, row = args
    img = (np.array(Image.open("../input/plant-pathology-2021-fgvc8/train_images/" + row.image)) / 255).reshape(-1, 3)
    return img.mean(axis=0), img.std(axis=0)

with mp.Pool() as pool:
    for m, s in tqdm(pool.imap_unordered(worker, df.iterrows()), total=len(df)):
        means.append(m)
        stds.append(s)
```

# Mean average per color channel

```python
np.vstack(means).mean(axis=0)
```

# Mean standard-deviation

```python
np.vstack(stds).mean(axis=0)
```
