```python
from sklearn.preprocessing import MultiLabelBinarizer
from tqdm.notebook import tqdm
import matplotlib.pyplot as plt
import tensorflow as tf
import albumentations
import pandas as pd
import numpy as np
import shutil
import os
from sklearn.model_selection import KFold, StratifiedKFold
```

```python
class CFG():
    img_size = 299
    seed = 42
    classes = [
        'complex', 
        'frog_eye_leaf_spot', 
        'powdery_mildew', 
        'rust', 
        'scab',
        'healthy']
    batch_size=32
```

```python
train_dir='../input/plant-pathology-2021-fgvc8/train.csv'
```

```python
test_img = '../input/plant-pathology-2021-fgvc8/test_images'
submission = pd.read_csv('../input/plant-pathology-2021-fgvc8/sample_submission.csv')
submission.head()
```

```python
submission['labels'] = submission['labels'].apply(lambda s: s.split(' '))
submission.head()
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
''' ImageDataGenerator參考網頁-> https://keras.io/zh/preprocessing/image/ '''
datagen = ImageDataGenerator(
    rescale = 1/255.,
    rotation_range = 15,
    width_shift_range=0.2, 
    height_shift_range=0.2,
    zoom_range = 0.2,
    shear_range=0.3, #（以弧度逆时针方向剪切角度）。
    horizontal_flip=True,
    vertical_flip=True,
    validation_split= 0.2,)
```

```python
test_generator = datagen.flow_from_dataframe(
                  submission,
                  directory = test_img,
                  x_col = 'image',
                  y_col = 'labels',
                  class_mode = "categorical",
                  batch_size=4,
                  target_size = (CFG.img_size,CFG.img_size),
                  color_mode="rgb",
                  shuffle = False,
                  seed = CFG.seed)
```

```python
import keras
def load_model():
    model_pre = keras.models.load_model("../input/n-2-0720/N_2_0720.h5")
    return model_pre
```

```python
model = load_model()
pred = model.predict(test_generator)
```

```python
perdict = (pred>0.36)
n_label = ['complex', 'frog_eye_leaf_spot', 'healthy', 'powdery_mildew', 'rust', 'scab']
answer = []

for i in range(perdict.shape[0]):
    temp = []
    for j, k in enumerate(n_label):
        if perdict[i, j]:
            temp.append(k)
    answer.append(temp)
    
answer = [' '.join(n) for n in answer]
```

```python
submission['labels'] = np.array(answer)
submission
```
