```python
from multiprocessing import Pool, cpu_count

import numpy as np
import tensorflow as tf
from PIL import Image
from tqdm import tqdm
```

```python
N = 100

f = "/kaggle/input/plant-pathology-2021-fgvc8/test_images/85f8cb619c66b863.jpg"

def parse_image(_):
    img = tf.io.read_file(f)
    img = tf.image.decode_jpeg(img, channels=3)
    img = tf.image.resize(
        img,
        (224, 224),
        method=tf.image.ResizeMethod.BICUBIC,
    ).numpy()

def _resize_tf(_):
    x = Image.open(f)
    x = tf.image.resize(
        np.array(x), 
        (224, 224),
        method=tf.image.ResizeMethod.BICUBIC
    ).numpy()


def _resize_PIL(_):
    x = Image.open(f)
    x = x.resize((224, 224), Image.BICUBIC)
    x = np.array(x)
```

```python
with Pool(cpu_count()) as pool:
    X = list(
        tqdm(
            pool.imap(
                parse_image,
                range(N),
            ),
            total=N,
            smoothing=0,
        )
    )
```

```python
for i in tqdm(range(N), smoothing=0):
    _resize_tf(i)
```

```python
for i in tqdm(range(N), smoothing=0):
    _resize_PIL(i)
```
