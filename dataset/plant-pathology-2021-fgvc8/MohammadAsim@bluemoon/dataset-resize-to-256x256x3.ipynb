```python
import numpy as np
import pandas as pd
import tensorflow as tf
import os
import concurrent.futures
import time
import cv2
```

```python
PATH = '../input/plant-pathology-2021-fgvc8/train_images'
namelist = os.listdir(PATH)
```

```python
IMSIZE = 256
def read_img(image, name):
    img = cv2.imread(image)
    img = cv2.resize(img, (IMSIZE, IMSIZE))
    return img, name
def prepare_dataset(namelist, path):
    start = time.time()
    names = namelist
    namelist = [os.path.join(path, ele) for ele in namelist]
    with concurrent.futures.ThreadPoolExecutor(max_workers = 16) as executor:
        i = 0
        for value, name in executor.map(read_img, namelist, names):
            i+=1
            print("\rFetching: [{}/{}]".format(i, len(namelist)), end="", flush=True)
            cv2.imwrite(name, value)
    print("\nExecution time: ",time.time() - start, "s")
```
