```python
import numpy as np
import pandas as pd

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader# For custom data-sets

import torchvision
import torchvision.transforms as T

import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import os
from PIL import Image
import seaborn as sns
```

```python
root_dir = '/kaggle/input/plant-pathology-2021-fgvc8/train_images'
out_dir = 'resize-plant-{}'
```

```python
def center_crop(img):
    """
    1:1的中心切图
    """
    w,h,c = img.shape
    if w > h:
        w0 = int((w-h)/2)
        w1 = w0 + h
        img = img[w0:w1,:]
    else:
        h0 = int((h-w)/2)
        h1 = h0 + w
        img = img[:,h0:h1]
    return img
```

```python
dsizes = [512, 256, 128, 64]

# 输出目录检测
for dsize in dsizes:
    resize_out_dir = out_dir.format(dsize)
    if not os.path.exists(resize_out_dir):
        os.mkdir(resize_out_dir)

# 图片resize
files = os.listdir(root_dir)
for file in files:
    img_path = os.path.join(root_dir, file)
    img = cv.imread(img_path)
    img = center_crop(img)
    
    # 输出图片
    for dsize in dsizes:
        resize_out_dir = out_dir.format(dsize)
        img_out_path = os.path.join(resize_out_dir, file)
        img_resized = cv.resize(img, dsize=(dsize, dsize))
        cv.imwrite(img_out_path, img_resized)
```
