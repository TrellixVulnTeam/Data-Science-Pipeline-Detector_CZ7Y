```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from numpy import asarray
```

```python
train = pd.read_csv("/kaggle/input/plant-pathology-2021-fgvc8/train.csv")
```

```python
for name, nn in zip(train.labels.unique(), range(1,len(train.labels))):
    print(f'{nn}. {name}')
```

```python
new_labels = []
labels = train.labels.unique().tolist()
for lab in labels:
    new_labels.append(lab.split(' '))
flat_list = [item for sub_list in new_labels for item in sub_list]
print(set(flat_list))
```

There are only **6 diseases**. But with the combination of two or three diseases It creates totoally 12 diseases.

```python
for name in train.labels.unique():
    img = mpimg.imread('/kaggle/input/plant-pathology-2021-fgvc8/train_images/'+train.loc[train.labels==name,'image'].values[0])
    imgplot = plt.imshow(img)
    plt.title(name +"-"+ str(img.shape)+"-"+str(img.dtype))
    plt.show()
```
