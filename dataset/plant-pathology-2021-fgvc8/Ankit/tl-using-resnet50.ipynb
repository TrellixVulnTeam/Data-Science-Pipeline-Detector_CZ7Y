<h1 style="font-family: Verdana; font-size: 24px; font-style: normal; font-weight: bold; text-decoration: none; text-transform: none; letter-spacing: 3px; background-color: #ffffff; color: navy;">ResNet as a rescuer&nbsp;&nbsp;&nbsp;&nbsp;</h1> 

After the celebrated victory of AlexNet at the LSVRC2012 classification contest, deep Residual Network was arguably the most groundbreaking work in the computer vision/deep learning community in the last few years. ResNet makes it possible to train up to hundreds or even thousands of layers and still achieves compelling performance.

Taking advantage of its powerful representational ability, the performance of many computer vision applications other than image classification have been boosted, such as object detection and face recognition.
Since ResNet blew people’s mind in 2015, many in the research community have dived into the secrets of its success, many refinements have been made in the architecture.

According to the universal approximation theorem, given enough capacity, we know that a feedforward network with a single layer is sufficient to represent any function. However, the layer might be massive and the network is prone to overfitting the data. Therefore, there is a common trend in the research community that our network architecture needs to go deeper.

Since AlexNet, the state-of-the-art CNN architecture is going deeper and deeper. While AlexNet had only 5 convolutional layers, the VGG network and GoogleNet (also codenamed Inception_v1) had 19 and 22 layers respectively.

However, increasing network depth does not work by simply stacking layers together. Deep networks are hard to train because of the notorious vanishing gradient problem — as the gradient is back-propagated to earlier layers, repeated multiplication may make the gradient infinitively small. As a result, as the network goes deeper, its performance gets saturated or even starts degrading rapidly.

The core idea of ResNet is introducing a so-called “identity shortcut connection” that skips one or more layers, as shown in the following figure:

The authors of argue that stacking layers shouldn’t degrade the network performance, because we could simply stack identity mappings (layer that doesn’t do anything) upon the current network, and the resulting architecture would perform the same. This indicates that the deeper model should not produce a training error higher than its shallower counterparts. They hypothesize that letting the stacked layers fit a residual mapping is easier than letting them directly fit the desired underlaying mapping. And the residual block above explicitly allows it to do precisely that.

Following this intuition, the authors of refined the residual block and proposed a pre-activation variant of residual block, in which the gradients can flow through the shortcut connections to any other earlier layer unimpededly. In fact, using the original residual block in, training a 1202-layer ResNet resulted in worse performance than its 110-layer counterpart.

<h1 style="font-family: Verdana; font-size: 24px; font-style: normal; font-weight: bold; text-decoration: none; text-transform: none; letter-spacing: 3px; background-color: #ffffff; color: navy;">Problem Statement&nbsp;&nbsp;&nbsp;&nbsp;</h1>

Apples are one of the most important temperate fruit crops in the world. Foliar (leaf) diseases pose a major threat to the overall productivity and quality of apple orchards. The current process for disease diagnosis in apple orchards is based on manual scouting by humans, which is time-consuming and expensive.

Although computer vision-based models have shown promise for plant disease identification, there are some limitations that need to be addressed. Large variations in visual symptoms of a single disease across different apple cultivars, or new varieties that originated under cultivation, are major challenges for computer vision-based disease identification. These variations arise from differences in natural and image capturing environments, for example, leaf color and leaf morphology, the age of infected tissues, non-uniform image background, and different light illumination during imaging etc.

Plant Pathology 2020-FGVC7 challenge competition had a pilot dataset of 3,651 RGB images of foliar disease of apples. For Plant Pathology 2021-FGVC8, we have significantly increased the number of foliar disease images and added additional disease categories. This year’s dataset contains approximately 23,000 high-quality RGB images of apple foliar diseases, including a large expert-annotated disease dataset. This dataset reflects real field scenarios by representing non-homogeneous backgrounds of leaf images taken at different maturity stages and at different times of day under different focal camera settings.

<b>Specific Objectives</b>

The main objective of the competition is to develop machine learning-based models to accurately classify a given leaf image from the test dataset to a particular disease category, and to identify an individual disease from multiple disease symptoms on a single leaf image.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
import random
import os
import glob
from tqdm.notebook import tqdm
import plotly
import plotly.express as ex
import plotly.graph_objs as go
import matplotlib.pyplot as plt
%matplotlib inline
# Preprocessing the data
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator,load_img
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import preprocess_input
import keras
from tensorflow.keras.optimizers import Adam, SGD
from keras.models import Model,load_model
from tensorflow.keras.layers import Dense, Dropout,Conv2D,Flatten,MaxPooling2D,BatchNormalization
from keras import backend as K
from tensorflow.keras.layers.experimental import preprocessing
from keras.callbacks import ReduceLROnPlateau,EarlyStopping
```

```python
# Look at the training, testing and validation data numbers
test_data = glob.glob('../input/plant-pathology-2021-fgvc8/test_images/*.jpg')
train_data = glob.glob('../input/plant-pathology-2021-fgvc8/train_images/*.jpg')
submission=pd.read_csv('../input/plant-pathology-2021-fgvc8/sample_submission.csv')
train_csv=pd.read_csv('../input/plant-pathology-2021-fgvc8/train.csv')

print(f"Training Set has: {len(train_data)} images")
print(f"Testing Set has: {len(test_data)} images")
```

### Preprocessing the images

```python
imageSize = 224
batchSize = 32
NUM_CLASSES = 6 ## since there are 6 distinct classes
```

### Prepare the image generator object

#### Input parameters of ImageDataGenerator for resnet50 pretrained model:
1. preprocessing_function : resnet50
2. rescale : for rescaling the input image 
3. shear range : Shear Intensity (Shear angle in counter-clockwise direction in degrees)
4. zoom_range : Range for random zoom
5. horizontal_flip : Randomly flip inputs horizontally
6. validation_split : validation set 

```python
trainGen = ImageDataGenerator(preprocessing_function = tf.keras.applications.resnet50.preprocess_input, 
                              rescale = 1./255,
                              shear_range = 0.2,
                              zoom_range = 0.2,
                              horizontal_flip = True,
                              vertical_flip=True,
                              brightness_range=(0.8, 1.2),
                              featurewise_center=True,
                              featurewise_std_normalization=True,
                              rotation_range=20,
                              validation_split = 0.2)
```

### Generated batches with augemented images from image generator class

```python
train_generator = trainGen.flow_from_dataframe(dataframe = train_csv,
                                               directory = train_image_dir,
                                               x_col = "image",
                                               y_col = "labels",
                                               class_mode= "categorical", 
                                               subset = "training", 
                                               target_size = (imageSize, imageSize),
                                               batch_size = batchSize,
                                               seed = 123)
val_generator = trainGen.flow_from_dataframe(dataframe = train_csv,
                                               directory = train_image_dir,
                                               x_col = "image",
                                               y_col = "labels",
                                               class_mode= "categorical", 
                                               subset = "validation", 
                                               target_size = (imageSize, imageSize),
                                               batch_size = batchSize,
                                               seed = 123)
```

```python
## lets check the type
type(val_generator) ## image generator object
```

```python
basemodel = tf.keras.applications.ResNet50(weights = "imagenet", include_top = False, pooling = "avg")
## We dont want to train our weight , we will use pre trained weights of resnet50 and remove the output layer
x = basemodel.output
x = Dense(128, activation='relu')(x)
x = BatchNormalization()(x)
top_dropout_rate = 0.2
x = Dropout(top_dropout_rate, name="top_dropout")(x)
x = Dense(64, activation='relu')(x)
predictions = Dense(NUM_CLASSES, activation='sigmoid')(x)
# this is the model we will train
model = Model(inputs=basemodel.input, outputs=predictions)
```

```python
basemodel.input### image of channel 3 and n*n dimension
```

```python
basemodel.output### flatten numpy array of size 2048
```

```python
model.summary()
```

### Create callback : Early stopping and ReduceLROnPlateau

```python
earlystop = EarlyStopping(monitor='val_loss',
                          min_delta=0.01,# threshold to consider as no change
                          patience=1,# stop if  epochs with no change
                          verbose=1,
                          mode='auto',
                          restore_best_weights= True)

# lr = ReduceLROnPlateau(monitor='val_loss', 
#                              factor=0.1, 
#                              patience=20, 
#                              verbose=1, 
#                              mode='auto', 
#                              min_delta=0.0001, 
#                              cooldown=0, 
#                              min_lr=0.0001)
```

```python
opt = keras.optimizers.Adam(lr=0.001, decay=1e-4)
model.compile(optimizer = opt, loss = "binary_crossentropy", metrics = ["accuracy"])
```

```python
## train for 3 epoch only
history = model.fit(train_generator, 
                    epochs=3, 
                    validation_data=val_generator, 
                    verbose=True, 
                    callbacks = earlystop)
```

### Save the trained Model

```python
model.save("resnet50_model.h5")
```

### Predictions on Test set

```python
test_df = pd.read_csv('../input/plant-pathology-2021-fgvc8/sample_submission.csv')
```

```python
test_image_dir = "/kaggle/input/plant-pathology-2021-fgvc8/test_images/"
test_generator = trainGen.flow_from_dataframe(dataframe = test_df,
                                               directory = test_image_dir,
                                               x_col = "image",
                                               y_col = "labels",
                                               class_mode= "categorical", 
                                               target_size = (imageSize, imageSize),
                                               batch_size = batchSize,
                                               seed = 123)
```

### Make prediction

```python
predResult = []
prediction = model.predict(test_generator).tolist()
```

```python
prediction## list of list
```

```python
train_generator.class_indices
```

### Generate predicted labels

```python
THRESHOLD=0.5
predResult = []
for img in prediction:
    predictionLabels = []
    index = 0
    for categoryScore in img:
        if categoryScore > THRESHOLD:
            predictionLabels.append(index)
        index += 1
    if not predictionLabels:
        predResult.append(np.argmax(prediction))
    else:
        predResult.append(predictionLabels)
```

```python
predResult
```

```python
label_map = (train_generator.class_indices)
final_label_map = dict((v,k) for k,v in label_map.items())
print(final_label_map)

for i in range(len(test_df)):
    labels = ""
    for idx in predResult[i]:
        labels += final_label_map[idx] + " "
    test_df.iloc[i, 1] = labels.strip()
```
