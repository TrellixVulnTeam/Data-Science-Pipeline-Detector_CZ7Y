```python
import numpy as np
import pandas as pd
import os
import fastai
from fastai.vision.all import *
from pathlib import Path
```

```python
path = Path('../input/plant-pathology-2021-fgvc8')
```

```python
test = pd.DataFrame(columns=['image', 'labels'])
test
```

```python
def get_x(r): return path/'train_images'/r['image']
def get_y(r): return r['labels'].split(' ')
```

```python
import pathlib
plt = platform.system()
if plt == 'Linux': pathlib.WindowsPath = pathlib.PosixPath
```

```python
learn_inf = load_learner('../input/pp2021model/export2.pkl')
learn_inf.metrics = partial(accuracy_multi, thresh=0.45)
```

```python
path = Path('../input/plant-pathology-2021-fgvc8/test_images/')
with learn_inf.no_bar(), learn_inf.no_logging():
    for filename in os.listdir(path):
            if filename.endswith('.jpg'):
                image = Image.open('../input/plant-pathology-2021-fgvc8/test_images/'+str(filename))
                image = np.asarray(image)
                prediction = learn_inf.predict(image)
                df = pd.DataFrame([[filename, ' '.join(prediction[0])]],
                                    columns=['image', 'labels'])
                test = test.append(df)
        
```

```python
test.to_csv(r'./submission.csv', index = False)
```
