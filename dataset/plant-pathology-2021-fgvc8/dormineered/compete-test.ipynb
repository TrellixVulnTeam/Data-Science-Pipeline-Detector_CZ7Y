```python
from PIL import Image
from torchvision import transforms as T
from tqdm import tqdm
import os
imgResize = T.Resize(size=684,interpolation=3)
def resize_images(path_in:str, path_out:str, filenames:list):
    if not os.path.isdir(path_out): os.mkdir(path_out)
    for filename in tqdm(filenames):
        image = Image.open(os.path.join(path_in, filename))
        if image.size[1]>image.size[0]:
            print(image.size)
            print('rot'+str(filenames))
            image = image.transpose(Image.ROTATE_90)
        image_resized = imgResize(img=image)
        save_to = os.path.join(path_out, filename)
        image_resized.save(save_to,quality=95)
```

```python
root_dir = '../input/plant-pathology-2021-fgvc8/train_images'
out_dir = './resize-plant-684'
resize_images(root_dir,out_dir,os.listdir(root_dir))
```
