```python
import glob

import matplotlib.pyplot as plt
from matplotlib.image import imread

import pandas as pd
# do not truncate dataframe data when displaying
pd.set_option('display.max_colwidth', None)
```

```python
# define constants and see what file structure kaggle defines
base_dir = "/kaggle/input/plant-pathology-2021-fgvc8/"
train_dir = base_dir + "train_images/"
glob.glob(base_dir + "*")
```

```python
# read the train.csv file into a dataframe
input_df = pd.read_csv(base_dir + "train.csv")
input_df['image'] = input_df['image'].apply(lambda x: train_dir + x)
input_df
```

```python
# get a series from labels and count how many we have of each type
# https://pandas.pydata.org/docs/reference/api/pandas.Series.plot.html
split_labels_df = input_df.copy()
split_labels_df['labels'] = split_labels_df['labels'].apply(lambda x: x.split(' '))
split_labels_df['labels'].explode('labels').value_counts().sort_values(ascending=True).plot(kind='barh')
```

```python
# get a preview_df with a sample of sample_count images for each label
sample_count = 3
preview_df = input_df.groupby('labels').head(sample_count).sort_values(by=['labels'])

# for each label display a subplot with sample_count images
for group_name, df_group in preview_df.groupby('labels'):
    figure, axes = plt.subplots(nrows=1, ncols=sample_count, figsize=[20, 20])
    for row_index, row in df_group.reset_index().iterrows():
        image = plt.imread(row['image'])
        axes[row_index].imshow(image)
        axes[row_index].set_title(group_name + str(row_index))
        axes[row_index].axis('off')
    plt.show()
```
