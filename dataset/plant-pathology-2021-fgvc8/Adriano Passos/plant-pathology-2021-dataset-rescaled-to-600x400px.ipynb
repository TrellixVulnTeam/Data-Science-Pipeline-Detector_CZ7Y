```python
from fastai.vision.all import *
```

## Copying the metadata files

```python
from shutil import copyfile
for f in Path('/kaggle/input/plant-pathology-2021-fgvc8').ls():
    if f.is_file():
        copyfile(f, Path('.')/f.name)
```

## Creating directory

```python
train_folder = Path('/kaggle/input/plant-pathology-2021-fgvc8/train_images')
train_folder.ls()
out_folder = Path('/kaggle/temp/train_images')
out_folder.mkdir(exist_ok=True, parents=True)
```

## Listing all images inside train directory

```python
files = get_image_files(train_folder)
files
```

## Creating and testing a pipeline to process the image

```python
from torchvision.transforms import Resize
```

```python
file = files[0]
pipe = Pipeline([PILImage.create, Resize(400)])
img = pipe(file)
show_image(img)
img.shape, out_folder/file.name
```

## Processing all the images in paralell and saving to the output directory

```python
def pre_process(file):    
    img = pipe(file)
    img.save(out_folder/file.name)
    img.close
```

```python
from joblib import Parallel, delayed  
from tqdm.auto import tqdm
out = Parallel(n_jobs=2)(delayed(pre_process)(file) for file in tqdm(files))
```
