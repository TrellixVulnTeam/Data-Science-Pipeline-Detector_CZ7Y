# Resizing the whole dataset

```python
import os
from tqdm.auto import tqdm
from joblib import Parallel, delayed
import multiprocessing
```

```python
n_jobs = multiprocessing.cpu_count()
n_jobs
```

```python
out_path = Path('train')
os.makedirs(out_path, exist_ok = True)
```

```python
train_path = Path('../input/plant-pathology-2021-fgvc8/train_images')
train_imgs = get_image_files(train_path)
train_imgs
```

```python
def resize_img_file(file, out_path):
    img = Image.open(file)
    img.thumbnail((size, size))
    img.save(out_path/file.name)
```

```python
_ = Parallel(n_jobs=n_jobs)(delayed(resize_img_file)(file, out_path) for file in tqdm(train_imgs))
```
