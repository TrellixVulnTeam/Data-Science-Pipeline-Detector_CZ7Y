# Exploratory Data Analysis

```python
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
plt.rcParams["figure.figsize"] = (14, 8)
sns.set_theme(context="notebook", style="whitegrid")
```

## Config

```python
# file paths
DATA_DIR = Path("..", "input", "tabular-playground-series-feb-2021")

# data
TRAIN_DATA = DATA_DIR / "train.csv"

# columns in the data
INDEX_COL = "id"

TARGET_COL = "target"

# random state
RANDOM_STATE = 42
```

## Loading the data

```python
df = pd.read_csv(TRAIN_DATA, index_col=INDEX_COL)
df.info()
```

## Pandas profiling report

```python
from pandas_profiling import ProfileReport

profile = ProfileReport(df)
profile
```

## Separate features from target

```python
y = df[TARGET_COL]
X = df.drop(TARGET_COL, axis=1)

numeric_dtypes = ["int64", "float64"]
categorical_df = X.select_dtypes(exclude=numeric_dtypes)
numeric_df = X.select_dtypes(include=numeric_dtypes)
```

## Handling missing values

```python
categorical_df = categorical_df.fillna(np.squeeze(categorical_df.mode()))
numeric_df = numeric_df.fillna(numeric_df.mean())
X = categorical_df.join(numeric_df)
any(X.isna().sum() > 0)
```

## Encoding categorical features

```python
for col in categorical_df.columns:
    X[col], _ = X[col].factorize()

X[categorical_df.columns].info()
```

## Mutual information

```python
# sampling
y_sample = y.sample(frac=0.8, random_state=RANDOM_STATE)
X_sample = X[X.index.isin(y_sample.index)]
```

```python
from sklearn.feature_selection import mutual_info_regression

discrete_features = X.columns.isin(categorical_df.columns)
mi_scores = mutual_info_regression(
    X_sample, y_sample, discrete_features=discrete_features, random_state=RANDOM_STATE
)

mi_scores = pd.Series(mi_scores, name="MI Scores", index=X.columns)
mi_scores = mi_scores.sort_values(ascending=False)
```

```python
fig, ax = plt.subplots()

sns.barplot(x=mi_scores.values, y=mi_scores.index, ax=ax, color="tab:blue")
ax.set_title("Mutual information scores")
plt.show()
```

```python
mutual_info = mi_scores[mi_scores > 0]
print("Informative features: {}".format(mutual_info.index.to_list()))
```
