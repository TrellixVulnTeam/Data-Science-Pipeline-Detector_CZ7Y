## Load in the data

```python
DATA_PATH = '/kaggle/input/tabular-playground-series-feb-2021/'
train_df = pd.read_csv(DATA_PATH + 'train.csv')
test_df = pd.read_csv(DATA_PATH + 'test.csv')
submission = pd.read_csv(DATA_PATH + 'sample_submission.csv')
```

```python
train_df.head()
```

## Encode categorical columns

```python
train_df.dtypes
```

```python
cat_features = [
    "cat0", "cat1", "cat2", "cat3", "cat4", "cat5", "cat6", "cat7", 
    "cat8", "cat9"
]

cont_features = [
    "cont0", "cont1", "cont2", "cont3", "cont4",
    "cont5", "cont6", "cont7", "cont8", "cont9", "cont10", 
    "cont11", "cont12", "cont13"
]
```

```python
from fastai.tabular.all import * 
y_names = 'target'
procs = [Categorify, FillMissing, Normalize]
dls = TabularDataLoaders.from_df(train_df, path = DATA_PATH, y_names=y_names, cat_names=cat_features, cont_names=cont_features, procs=procs)
```

```python
dls.show_batch()
```

```python
learn = tabular_learner(dls, metrics=rmse)
learn.fine_tune(30)
```

```python
test_df.head()
```

```python
test_dls = learn.dls.test_dl(test_df)
preds, _ = learn.get_preds(dl=test_dls)
preds.numpy().shape
```

```python
submission.head()
```

```python
submission.target = preds.numpy()
```

```python
submission.head()
```

```python
submission.to_csv('tabular_learner.csv', index=False)
```
