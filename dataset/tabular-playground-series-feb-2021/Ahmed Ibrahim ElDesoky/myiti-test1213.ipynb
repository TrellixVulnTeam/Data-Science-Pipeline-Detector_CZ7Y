```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns

#plt.style.use('seaborn') #ggplot

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_error
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
train = pd.read_csv(r"../input/tabular-playground-series-feb-2021/train.csv")
test = pd.read_csv(r"../input/tabular-playground-series-feb-2021/test.csv")
```

```python
train=pd.get_dummies(train)
```

```python
train.info()
```

```python
X =train.drop(['id', 'target', 'cat6_G'], axis=1)
Y = pd.DataFrame(train['target'])
```

```python
X_train, x_test, Y_train, y_test = train_test_split(X, Y, test_size=0.1,random_state=60)
```

```python
scaler_x = StandardScaler().fit(X_train)
scaler_y = StandardScaler().fit(Y_train)
```

```python
x_train_sc = scaler_x.transform(X_train)
x_test_sc = scaler_x.transform(x_test)
```

```python
y_train_sc = scaler_y.transform(Y_train)
y_test_sc = scaler_y.transform(y_test)
```

```python
lr_model = LinearRegression()
lr_model.fit(x_train_sc, y_train_sc)
y_pred_sc = lr_model.predict(x_test_sc)
```

```python
mae = mean_absolute_error(y_test_sc, y_pred_sc)
rmse = np.sqrt(mean_squared_error(y_test_sc, y_pred_sc))

print('MAE = ', mae)
print('RMSE = ', rmse)
```

```python
y_test_inv = scaler_y.inverse_transform(y_test_sc.reshape(-1,1))
y_pred_inv = scaler_y.inverse_transform(y_pred_sc.reshape(-1,1))

y_test_inv

actual_mae = mean_absolute_error(y_test_inv, y_pred_inv)
actual_rmse = np.sqrt(mean_squared_error(y_test_inv, y_pred_inv))

print('MAE = ', actual_mae)
print('RMSE = ', actual_rmse)
```

```python
test = pd.get_dummies(test)
test
```

```python
test_data = test.drop(['id'], axis=1)
data_sc = scaler_x.transform(test_data)
```

```python
test_predictions_sc = lr_model.predict(data_sc)
```

```python
test_predictions = scaler_y.inverse_transform(test_predictions_sc.reshape(-1,1))
test_predictions
```

```python
submission = pd.DataFrame(test.id)
submission['target'] = test_predictions
submission
```

```python
submission.to_csv('MySubm.csv',index=False)
```
