```python
train = pd.read_csv("/kaggle/input/tabular-playground-series-feb-2021/train.csv", index_col='id')
test = pd.read_csv("/kaggle/input/tabular-playground-series-feb-2021/test.csv", index_col='id')
sample = pd.read_csv("/kaggle/input/tabular-playground-series-feb-2021/sample_submission.csv")
```

```python
for i in train.columns:
    if train[i].dtype=='object': 
        le = LabelEncoder()
        le.fit(list(train[i].values) + list(test[i].values))
        train[i] = le.transform(train[i].values)
        test[i] = le.transform(test[i].values)
        
display(train.head())
display (test.head())
```

```python
x = train.drop(['target'], axis=1)
y = train['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, train_size=0.80)
```

```python
clf = RandomForestRegressor(n_estimators=200, max_depth=7, n_jobs=-1)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
y_pred_test = clf.predict(test)
```

```python
submission = pd.DataFrame({
        "id": sample['id'],
        "target": y_pred_test
    })

submission.to_csv('submission.csv',index = False)
```
