```python
train = pd.read_csv('/kaggle/input/tabular-playground-series-feb-2021/train.csv')
test = pd.read_csv('/kaggle/input/tabular-playground-series-feb-2021/test.csv')
```

```python
train
```

```python
reg = CascadeForestRegressor(predictor = 'lightgbm',partial_mode =True,n_estimators=1)
```

```python
y = train['target']
train.drop(['target','id'],axis=1,inplace=True)
test.drop(['id'],axis=1,inplace=True)
```

```python
for col in [col for col in train.columns if 'cat' in col]:
    train[col] = pd.factorize(train[col])[0]
    test[col] = pd.factorize(test[col])[0]
```

```python
train
```

```python
reg.fit(train.values,y.values.reshape(-1,1).ravel())
```

```python
def fast_rmse(targets,predictions):
    n = len(predictions)
    rmse = np.linalg.norm(predictions - targets) / np.sqrt(n)
    return rmse
```

```python
fast_rmse(reg.predict(train.values),y)
```
