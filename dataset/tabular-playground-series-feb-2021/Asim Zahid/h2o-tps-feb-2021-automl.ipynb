# H2O

```python
h2o.init(max_mem_size='12G', nthreads=4)
```

# Here

```python
train = h2o.import_file("../input/tabular-playground-series-feb-2021/train.csv")
test = h2o.import_file("../input/tabular-playground-series-feb-2021/test.csv")

x = train.columns
y = "target"
#x = x.drop("target").column
#x.remove(y)

#train = h2o.H2OFrame(train)
```

```python
# @tunguz
object_columns = x[:10]
for f in object_columns:
    train[f] = train[f].asfactor()
    test[f] = test[f].asfactor()
```

```python
# Run AutoML for 20 base models (limited to 1 hour max runtime by default)
aml = H2OAutoML(max_models=20, seed=1234, max_runtime_secs=10800)
aml.train(x=x, y=y, training_frame=train)
```

```python
# View the AutoML Leaderboard
lb = aml.leaderboard
lb.head(rows=lb.nrows)  # Print all rows instead of default (10 rows)
```

```python
nuts = aml.leader
nuts
```

```python
preds = aml.predict(test).as_data_frame()
preds.head()
```

```python
sub_df = pd.read_csv("../input/tabular-playground-series-feb-2021/sample_submission.csv")
```

```python
sub_df['target'] = preds['predict']
sub_df.to_csv('submission_h2o.csv', index=False)
sub_df.head()

sub_automl = preds['predict']
```
