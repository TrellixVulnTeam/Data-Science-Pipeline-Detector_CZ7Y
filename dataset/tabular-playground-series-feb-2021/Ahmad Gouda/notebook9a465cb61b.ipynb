### Encoding Features

```python
df = pd.get_dummies(df)
df.columns
```

```python
X = df.drop(["id", "target", "cat6_G"], axis=1)
Y = pd.DataFrame(df["target"])
```

```python
X_df, x_dfT, Y_df, y_dfT = train_test_split(X, Y, test_size=0.1,random_state=42)
```

```python
print("X_train shape: ", X_df.shape)
print("X_train shape: ", Y_df.shape)
print("X_train shape: ", x_dfT.shape)
print("X_train shape: ", y_dfT.shape)
```

### 5. Data Preparation

```python
scaler_x = StandardScaler().fit(X_df)
scaler_y = StandardScaler().fit(Y_df)
```

```python
x_df_sc = scaler_x.transform(X_df)
x_dfT_sc = scaler_x.transform(x_dfT)
```

```python
y_df_sc = scaler_y.transform(Y_df)
y_dfT_sc = scaler_y.transform(y_dfT)
```

### Model


```python
lr_model = LinearRegression()
lr_model.fit(x_df_sc, y_df_sc)
y_pred_sc = lr_model.predict(x_dfT_sc)
```

```python
mae = mean_absolute_error(y_dfT_sc, y_pred_sc)
rmse = np.sqrt(mean_squared_error(y_dfT_sc, y_pred_sc))

print('MAE = ', mae)
print('RMSE = ', rmse)
```

```python
y_dfT_inv = scaler_y.inverse_transform(y_dfT_sc.reshape(-1,1))
y_pred_inv = scaler_y.inverse_transform(y_pred_sc.reshape(-1,1))

y_dfT_inv

actual_mae = mean_absolute_error(y_dfT_inv, y_pred_inv)
actual_rmse = np.sqrt(mean_squared_error(y_dfT_inv, y_pred_inv))

print('Actual MAE = ', actual_mae)
print('Actual RMSE = ', actual_rmse)
```

```python
dfT = pd.get_dummies(dfT)
dfT
```

```python
test_data = dfT.drop(['id'], axis=1)
data_sc = scaler_x.transform(test_data)
```

```python
test_predictions_sc = lr_model.predict(data_sc)
```

```python
test_predictions = scaler_y.inverse_transform(test_predictions_sc.reshape(-1,1))
test_predictions
```

```python
submission = pd.DataFrame(dfT.id)
submission['target'] = test_predictions
submission
```

```python
submission.to_csv('Ml_Mine.csv',index=False)
```
