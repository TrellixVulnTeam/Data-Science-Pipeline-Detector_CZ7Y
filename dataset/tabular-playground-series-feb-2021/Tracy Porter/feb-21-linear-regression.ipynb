Define y

```python
y = train.target
y
```

Combine train and test

```python
X_tot = train.drop('target', axis=1).append(test)
X_tot
```

Check for null values

```python
X_tot.isnull().sum()
```

Ordinal encode catecorical columns

```python
#ordinal encoder
from sklearn import preprocessing
from sklearn.preprocessing import OrdinalEncoder
enc = OrdinalEncoder()

X_tot.cat0 = enc.fit_transform(X_tot.cat0.values.reshape(-1,1))
X_tot.cat1 = enc.fit_transform(X_tot.cat1.values.reshape(-1,1))
X_tot.cat2 = enc.fit_transform(X_tot.cat2.values.reshape(-1,1))
X_tot.cat3 = enc.fit_transform(X_tot.cat3.values.reshape(-1,1))
X_tot.cat4 = enc.fit_transform(X_tot.cat4.values.reshape(-1,1))
X_tot.cat5 = enc.fit_transform(X_tot.cat5.values.reshape(-1,1))
X_tot.cat6 = enc.fit_transform(X_tot.cat6.values.reshape(-1,1))
X_tot.cat7 = enc.fit_transform(X_tot.cat7.values.reshape(-1,1))
X_tot.cat8 = enc.fit_transform(X_tot.cat8.values.reshape(-1,1))
X_tot.cat9 = enc.fit_transform(X_tot.cat9.values.reshape(-1,1))
```

```python
X_tot
```

Drop id number

```python
X_tot = X_tot.drop('id', axis=1)
X_tot
```

Normalise X_tot

```python
X_tot = (X_tot - X_tot.min()) / (X_tot.max() - X_tot.min())
X_tot
```

Split X_tot for training and testing

```python
rows = len(train)

X_Train = X_tot[: rows]
X_test = X_tot[rows :]
```

```python
X_Train
```

```python
X_test
```

Split X_train for training and validation

```python
from sklearn.model_selection import train_test_split

# Split into validation and training data
X_train, X_val, y_train, y_val = train_test_split(X_Train, y, test_size=0.05, random_state=1)
X_train.shape, X_val.shape, y_train.shape,y_val.shape, X_test.shape
```

Define model

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression().fit(X_train, y_train)
print(model.score(X_train, y_train))
```

Predict on validation set

```python
y_pred = model.predict(X_val)
print(model.score(X_val, y_val))
```

```python
from sklearn.metrics import mean_squared_error

rmse = mean_squared_error(y_val, y_pred, squared=False)
rmse
```

```python
df = pd.DataFrame({'Actual': y_val, 'Predicted':y_pred})
df
```

```python
plt.figure(figsize=(10,10))
plt.scatter(y_val, y_pred, c='crimson')
plt.yscale('log')
plt.xscale('log')

p1 = max(max(y_pred), max(y_val))
p2 = min(min(y_pred), min(y_val))
plt.plot([p1, p2], [p1, p2], 'b-')
plt.xlabel('Actual Values', fontsize=15)
plt.ylabel('Predictions', fontsize=15)
plt.axis('equal')
plt.show()
```

Predict on test set and submit

```python
final_labels = model.predict(X_test)

final_labels[final_labels < 0] = 0
final_labels
```

```python
final_result = pd.DataFrame({'id': test.id, 'target': final_labels})

final_result.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```
