```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

#### Load data...

```python
train = pd.read_csv("/kaggle/input/tabular-playground-series-feb-2021/train.csv")
test = pd.read_csv("/kaggle/input/tabular-playground-series-feb-2021/test.csv")
sample_submission = pd.read_csv('../input/tabular-playground-series-feb-2021/sample_submission.csv')
```

### Look at data...

```python
train = train.drop('id', axis = 1)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))

sns.boxplot(data = train);
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(train.corr(), annot= True);
```

```python
rfr = RandomForestRegressor(max_depth=5)
```

```python
categorical_cols = ['cat0', 'cat1', 'cat2', 'cat3', 'cat4', 'cat5', 'cat6', 'cat7','cat8', 'cat9']

cat_encoded_train = pd.get_dummies(train[categorical_cols])
```

```python
train = pd.concat([train, cat_encoded_train], axis= 1)
train = train.drop(categorical_cols, axis =1)
```

```python
train.columns
```

```python
from sklearn.model_selection import train_test_split

x = train.drop('target', axis = 1)
y = train.target

x_train, x_test, y_train, y_test = train_test_split(x,y,train_size = 0.7)
```

```python
import time
```

```python
%time model = rfr.fit(x_train, y_train)
```

```python
preds = rfr.predict(x_test)
```

```python
from sklearn.metrics import *

mean_squared_error(y_test, preds)
```

```python
mean_absolute_error(y_test, preds)
```
