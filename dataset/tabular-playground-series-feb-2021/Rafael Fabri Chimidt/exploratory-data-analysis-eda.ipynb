# Plots

![](https://image.freepik.com/free-vector/data-extraction-concept-illustration_114360-4766.jpg)

Let's build some plots!!!

We can start building histograms for all continuous variables.

```python
#plot histogram
fig, ax = plt.subplots(figsize = (12,8))

df_train.hist(ax = ax, bins = 20)

fig.tight_layout();
```

Well 

```python
#target histogram
fig, ax = plt.subplots(figsize = (12,8))

sns.distplot(x=df_train['target'], ax=ax, bins=100)

fig.tight_layout();
```

The target variable is bimodal.

Let's see if there are outliers in the dataset with boxplots. 

```python
#plot boxplot
fig, ax = plt.subplots(figsize = (12,8))

sns.boxplot(data = df_train, ax=ax)

fig.tight_layout();
```

Wow, there are many outliers in the target variable. Besides that, there are outliers in cont0, cont2, cont6, cont8.

Below is the interquartile range calculation.

```python
#Outliers target
#quartis target
Q1 = df_train['target'].quantile(0.25)
Q3 = df_train['target'].quantile(0.75)

#
QR = Q3 - Q1
max_iqr = Q3 + 1.5*QR
min_iqr = Q1 - 1.5*QR
print('maximium value =', max_iqr, '\nminimum value = ', min_iqr)
print('numbers outliers = ', round( ( ( (
    df_train['target']>max_iqr).sum() + (df_train['target']<min_iqr).sum() ).sum() / df_train.shape[0] ) * 100, 3), '%')

#plotando boxplot cont7
sns.boxplot(x = df_train['target'], data = df_train);
```

```python
#plot correlation graph
fig, ax = plt.subplots(figsize = (12,8))

sns.heatmap(df_train.corr(), ax = ax, annot=True, fmt='.0g', cmap = 'YlGnBu')

fig.tight_layout();
```
