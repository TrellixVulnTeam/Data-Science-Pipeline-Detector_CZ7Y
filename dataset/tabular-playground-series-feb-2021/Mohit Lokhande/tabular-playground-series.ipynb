```python
# Outlier detection
sns.boxplot(data=train_df['target'])
```

```python
# Removing Outlier
def out_liers(dss):
    for col in dss.columns:
        Q1 = dss[col].quantile(0.25)
        Q3 = dss[col].quantile(0.75)
        IQR = Q3 - Q1
        median = dss[col].median()
        dss.loc[((dss[col] < Q1 - 1.5* IQR) | (dss[col] > Q3 + 1.5*IQR)), col] = median
    return dss
```

```python
#Replace Outliers
train_df = out_liers(train_df)
test_df = out_liers(test_df)
```

```python
X = train_df.drop(['target'], axis=1)
y = train_df.target
```

```python
X.info()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=55)
```

```python
from sklearn.ensemble import RandomForestRegressor


regressor = RandomForestRegressor(n_estimators = 78, random_state = 0)
regressor = regressor.fit(X_train,y_train)
```

```python
 y_pred = regressor.predict(X_test)
```

```python
from sklearn import metrics
print("Mean absolute error (MAE):", metrics.mean_absolute_error(y_test,y_pred))
print("Mean square error (MSE):", metrics.mean_squared_error(y_test,y_pred))
```

```python
print("Root Mean Square Error :", np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
print("r2_Square",metrics.r2_score(y_test,y_pred))
```
