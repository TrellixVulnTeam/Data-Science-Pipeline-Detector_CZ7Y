```python
'''
The following code is for the CatBoosting
Created by - ANALYTICS VIDHYA
'''

# importing required libraries
import pandas as pd
import numpy as np
from catboost import CatBoostRegressor
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
# read the train and test dataset
train_data = pd.read_csv('/kaggle/input/tabular-playground-series-feb-2021/train.csv')
test_data = pd.read_csv('/kaggle/input/tabular-playground-series-feb-2021/test.csv')

# shape of the dataset
print('Shape of training data :',train_data.shape)
print('Shape of testing data :',test_data.shape)

# Now, we have used a dataset which has more categorical variables
# hr-employee attrition data where target variable is Attrition 

# seperate the independent and target variable on training data
train_x = train_data.drop(columns=['target'],axis=1)
train_y = train_data['target']

# seperate the independent and target variable on testing data
X_train,X_validation,y_train,y_validation = train_test_split(train_x,train_y)

# find out the indices of categorical variables
categorical_var = np.where(train_x.dtypes != np.float)[0]
print('\nCategorical Variables indices : ',categorical_var)

model = CatBoostRegressor(iterations=50)

# fit the model with the training data
model.fit(X_train,y_train,cat_features = categorical_var,eval_set=(X_validation, y_validation),plot=True)
print('\n Model Trainied')
```

```python
# predict the target on the train dataset
predict_train = model.predict(X_train)
print('\nTarget on train data',predict_train) 

# Accuray Score on train dataset
from sklearn.metrics import mean_squared_error
from math import sqrt

rms = sqrt(mean_squared_error(y_train, predict_train))

print('\RMSE on train dataset : ', rms)
```

```python
# predict the target on the test dataset
predict_eval = model.predict(X_validation)
print('\nTarget on evaluation split',predict_eval) 

# Accuray Score on test dataset
rms = sqrt(mean_squared_error(y_validation, predict_eval))

print('\RMSE on evaluation split : ', rms)
```

```python
test = pd.read_csv('/kaggle/input/tabular-playground-series-feb-2021/test.csv')
```

```python
preds = model.predict(test)
```

```python
test['target']=preds
test=test[['id','target']]
test.to_csv('submission.csv',index=False)
```
