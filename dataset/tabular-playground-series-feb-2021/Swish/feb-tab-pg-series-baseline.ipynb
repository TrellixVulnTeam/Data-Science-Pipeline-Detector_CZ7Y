Before I start EDA, I created a simple linear regression model in order to find a baseline model for this competition.

```python
train = pd.read_csv("../input/tabular-playground-series-feb-2021/train.csv")
test = pd.read_csv("../input/tabular-playground-series-feb-2021/test.csv")
allrows = pd.concat([train, test], axis = 0)
# check for NAs
allrows.isna().sum()
```

Since we have no NAs (except those we found in the target column in the test set), we can start modelling right away. First, we enode the categorical variables (now represented as strings) as integers so that we can include them in our regression model. 
We also remove ID and the target variable from the dataset in order to use the target variable as our dependent variable later in regression. Removing it now makes handling the joined dataset easier, as it has the same columns.

```python

edited_train = train.drop(["id", "target"], axis = 1)
edited_test = test.drop(["id"], axis = 1)

allrows = pd.concat([edited_train, edited_test], axis = 0)

columns = edited_train.columns
# the first through tenth columns in the dataset are categorical
cat_columns = columns[0:10]
# get dummy variables in order to model categorical features
dummies = pd.get_dummies(allrows, prefix = cat_columns)

# after we are done with our short editing, we split the datasets into test and train set again
edited_train = dummies[:300000]
edited_test = dummies[300000:]

edited_train
```

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()


# get dummy variables in order to model categorical features
Y = train.target

# fit our model
model = linreg.fit(edited_train, Y)
```

Finally, we predict the target variable and export our predictions to a csv

```python
baseline_predictions = model.predict(edited_test)
```

```python
predictions = pd.DataFrame(test["id"])
predictions["target"] = baseline_predictions
predictions.reset_index()
predictions.to_csv("./submission.csv", index = False)
```
