```python
import pandas as pd
import numpy as np
import os
from pathlib import Path
from fastai.tabular.all import *
```

```python
path = Path('../input/tabular-playground-series-feb-2021')
os.listdir(path)
```

```python
train = pd.read_csv(path / 'train.csv')
test = pd.read_csv(path / 'test.csv')
sub = pd.read_csv(path / 'sample_submission.csv')
```

```python
train.head()
```

```python
train.columns
```

```python
cat_names = []
cont_names = []
for col in train.columns:
    if col.startswith('cat'):
        cat_names.append(col)
    elif col.startswith('cont'):
        cont_names.append(col)
```

```python
cat_names
```

```python
cont_names
```

```python
y_names = 'target'
train = train.drop(columns = ['id'])
```

```python
procs = [Categorify, FillMissing, Normalize]
```

```python
data = TabularDataLoaders.from_df(train, path = path, procs = procs, cat_names = cat_names, cont_names = cont_names, y_names = y_names )
```

```python
data.show_batch()
```

```python
learn = tabular_learner(data)
```

```python
learn.fit_one_cycle(10)
```

```python
learn.validate()
```

```python
test.head()
```

```python
test = test.drop(columns = ['id'])
```

```python
tst_dl = learn.dls.test_dl(test)
pred, _ = learn.get_preds(dl = tst_dl)
```

```python
sub['target'] = pred.squeeze()
sub.to_csv('submission.csv', index=False)
sub
```
