- There are many techniques for handling obvious features,


        - Here I prefer to use a one hot and label encoding technique

- ['cat0','cat1','cat2'] having only two feature so we performe one hot encoding technique on it,
- If you want to explore One Hot encoding please take of my [notebook](https://www.kaggle.com/rushikeshlavate/check-how-to-perform-onehot-encoding-technique)

```python
for i in train_data.columns[1:4]:
    train_data[i] = pd.get_dummies(train_data[i],drop_first=True,columns=[i])
    train_data.rename(columns={'B':i},inplace=True)
    train_data.head()
```

```python
for i,col in enumerate(train_data.columns[4:11],1):
    print(col,"=",list(train_data[col].value_counts().index),"\n=============")
```

```python
cat3_dict = {'C':1, 'A':2, 'D':3, 'B':4}
cat4_dict = {'B':1, 'A':2, 'C':3, 'D':4}
cat5_dict = {'B':1, 'D':2, 'C':3, 'A':4}
cat6_dict = {'A':1, 'B':2, 'C':3, 'D':4, 'I':5, 'E':6, 'H':7, 'G':8}
cat7_dict = {'E':1 ,'D':2, 'B':3, 'G':4, 'F':5, 'A':6, 'C':7, 'I':8}
cat8_dict = {'C':1, 'E':2, 'G':3, 'A':4, 'D':5, 'F':6, 'B':7}
cat9_dict = {'F':1, 'I':2, 'L':3, 'H':4, 'K':5, 'A':6, 'G':7, 'M':8, 'J':9, 'O':10, 'N':11, 'B':12, 'C':13, 'D':14, 'E':15}
cat_dict = [cat3_dict,cat4_dict,cat5_dict,cat6_dict,cat7_dict,cat8_dict,cat9_dict]
cat_dict
```

```python
cat_dict = [cat3_dict,cat4_dict,cat5_dict,cat6_dict,cat7_dict,cat8_dict,cat9_dict]
for i,j in zip(train_data.columns[4:11],cat_dict):
        train_data[i]=(train_data[i].map(j))     
```

<h3 style="color:#7F525D"> - Building model

```python
from sklearn.model_selection import train_test_split
import xgboost as xgb
```

```python
x = train_data.drop(['id','target'],axis=1)
y=  train_data['target']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=18)
Classifier  = xgb.XGBRegressor()
Classifier.fit(x_train,y_train)
```

```python
demo = test_data.drop('id',axis=1)
```

```python
y_pred = Classifier.predict(demo)
y_pred
```

```python
submition = pd.DataFrame({
    'id':test_data['id'],
    'target':y_pred
})
```
