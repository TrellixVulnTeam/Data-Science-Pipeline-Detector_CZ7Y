```python
def plot_whole_features(df):
    columns_num = df.columns
    sns.set_context("paper", font_scale=2)
    fig, axes = plt.subplots(nrows=round(len(columns_num)/2), ncols=2, figsize=(40, 100))
    for ax, column in zip(axes.flatten(),columns_num):
        sns.histplot(df[column], ax=ax )
        sns.set(font_scale=1)
```

```python
plot_whole_features(train[train.columns[12:25]])
```

```python
plot_whole_features(train[train.columns[12:25]])
```

```python
import plotly.figure_factory as ff
def hist_plotly(df, col,bin_size=4,color = "slategray"):
    colors = [color]
    fig = ff.create_distplot([df[col].dropna()], [col],bin_size=bin_size,
                             curve_type='kde', # override default 'kde'
                             show_rug=False,
                             colors=colors)

    # Add title
    fig.update_layout(
                  template= "plotly_dark" , 
                  xaxis = dict(title = col.upper()))
    fig.show()
```

```python
hist_plotly(train, "target", bin_size = .02)
```

```python
def plot_feature_scatter(df1, df2, features):
    i = 0
    sns.set_style('whitegrid')
    plt.figure()
    fig, ax = plt.subplots(5, 3,figsize=(14, 24))

    for feature in features:
        i += 1
        plt.subplot(5, 3, i)
        plt.scatter(df1[feature], df2[feature], marker='+', color=['red'])
        plt.xlabel(feature, fontsize=9)
    plt.show();
```

```python
plot_feature_scatter(train[::15],test[::10], features=train.columns[12:25])
```

```python
def plot_feature_target_scatter(df1, features):
    i = 0
    sns.set_style('whitegrid')
    plt.figure()
    fig, ax = plt.subplots(5, 3,figsize=(14, 24))

    for feature in features:
        i += 1
        plt.subplot(5, 3, i)
        plt.scatter(df1[feature], df1['target'], marker='+', color='blue')
        plt.xlabel(feature, fontsize=9)
    plt.show();
```

```python
plot_feature_target_scatter(train[::15],features=train.columns[12:25] )
```
