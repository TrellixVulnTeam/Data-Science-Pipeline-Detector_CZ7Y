```python
df = pd.get_dummies(df ,columns= ['cat0', 'cat1', 'cat2', 'cat3', 'cat4', 'cat5', 'cat6', 'cat7',
       'cat8', 'cat9'])
df.head()
```

```python
df.columns
```

# **Spilt Data**

```python
from sklearn.model_selection import train_test_split
# select split parameter
x = df.drop(['target','id'], axis=1)
y = df.target
```

```python
# splitting into train and test sets for both parameters
x_train, x_test ,y_train , y_test = train_test_split(x,y, test_size=0.7, random_state= 42)
```

# **Models**

# 1.**Random Forest Regressor**

```python
from sklearn.ensemble import RandomForestRegressor


rfr = RandomForestRegressor()
```

```python
rfr.fit(x_train, y_train)
```

```python
rfr_preds = rfr.predict(x_test)

rfr_preds
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(y_test, rfr_preds)
```

```python
from sklearn.metrics import mean_absolute_error

mean_absolute_error(y_test, rfr_preds)
```

# 2.**KNeighbors Regressor**

```python
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor()
```

```python
knn.fit(x_train, y_train)
```

```python
knn_preds = knn.predict(x_test)

knn_preds
```

```python
mean_squared_error(y_test, knn_preds)
```

```python
mean_absolute_error(y_test, knn_preds)
```

# 3.**LinearRegression**

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
lr_preds = lr.predict(x_test)

lr_preds
```

```python
mean_squared_error(y_test, lr_preds)
```

```python
mean_absolute_error(y_test, lr_preds)
```
