```python
def print_score(m):
    res = [mean_absolute_error(m.predict(X_train), y_train),
           mean_absolute_error(m.predict(X_valid), y_valid),
           m.score(X_train, y_train), m.score(X_valid, y_valid)]
    if hasattr(m, 'oob_score_'): res.append(m.oob_score_)
    print(res)
```
