```python
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
data = pd.read_csv('/kaggle/input/pubg-finish-placement-prediction/train_V2.csv')
```

```python
data
```

```python
data.info()
```

```python
null_columns=data.columns[data.isnull().any()]
data[null_columns].isnull().sum()
```

**Quality** 
The column winPlacePerc has a NaN value

```python

data[data['winPlacePerc'].isnull()]
```

**The null value is in the 2744604 th Row.**

```python
data.drop(2744604, inplace=True)
```

**Removing The Null Value**

```python
sns.jointplot(x='winPlacePerc', y='kills', data=data, height=10, color="red")
plt.show()
```

```python
sns.jointplot(x='winPlacePerc', y='damageDealt', data=data, height=10, color="maroon")
plt.show()
```

**Killing and Dealing damage to opponents have a similar but moderate correlation to winning**

```python
killcat=data.copy()
killcat['killCategory'] = pd.cut(killcat['kills'], [-1, 0, 5, 10, 20, 60], labels=['0_kills','1-5_kills', '6-10_kills', '11-20_kills', '10+_kills'])
plt.figure(figsize=(10,6))
sns.boxplot(x='killCategory', y='winPlacePerc', data=killcat)
plt.show()
```

-**But If you do want to kill your way to victory, you better get more than 6 kills but try not to get too blood thirsty past 20 kills as it decreases your chances of winning**

```python
sns.jointplot(x='winPlacePerc', y='walkDistance', data=data, height=10, color="green")
plt.show()
```

**Distance Walked has a correlation with winning**

```python
sns.jointplot(x='winPlacePerc', y='rideDistance', data=data, height=10,  color="blue")
plt.show()
```

**Riding Vehicles for longer distances have mild correlation with winning**

```python
data['healsandboosts'] = data['heals'] + data['boosts']
data[['heals', 'boosts', 'healsandboosts']].tail()
```

**Creating a new column to combine player supporting features together.**

```python
sns.jointplot(x='winPlacePerc', y='healsandboosts', data=data, height=10,  color="lime")
plt.show()
```

# **Healing and Boosting yourself is a definite key to winning more games.**

```python
sns.jointplot(x='winPlacePerc', y='weaponsAcquired', data=data, height=10,  color="black")
plt.show()
```

**Acquiring multiple weapons will not help a player in winning more games.**

```python
sns.jointplot(x='winPlacePerc', y='revives', data=data, height=10,  color="#34495e")
plt.show()
```

```python
revcat=data.copy()
revcat['revCategory'] = pd.cut(revcat['revives'], [-1, 0, 2, 5, 10, 40], labels=['0_revives','1-2_revives', '3-5_revives', '6-10_revives', '10+_revives'])
plt.figure(figsize=(10,6))
sns.boxplot(x='revCategory', y='winPlacePerc', data=revcat)
plt.show()
```

**Reviving your teammates decrease you chances of ending up at the bottom but not that influential to winning. Chances are if they were bad enough to go down in the first place they won't be of much help anyway LOL.** Also who revived his teammates ~40 times? Sir are you a god?
