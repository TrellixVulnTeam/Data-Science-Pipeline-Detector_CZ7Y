# **Importing Libraries**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder
```

# **Loading Datasets**

Read data from csv file in form dataframe by using 'pandas' library.

```python
train = pd.read_csv('../input/pubg-finish-placement-prediction/train_V2.csv',nrows=100000)
test = pd.read_csv('../input/pubg-finish-placement-prediction/test_V2.csv',nrows=30000)
test_id = test['Id']
```

Use function info() for describe  train and train dataframe

```python
print(train.info())
```

Use function shape for describe size of train dataframe in array.

```python
print(train.shape)
```

Drop rows of train data if that row has missing value.

```python
train = train.dropna()
print(train.isnull().sum())
```

# **Label Encoding**

```python
Le1=LabelEncoder()
Le1.fit(train['Id'])
train['Encoded_Id']=Le1.transform(train['Id'])

Le2=LabelEncoder()
Le2.fit(train['groupId'])
train['Encoded_groupId']=Le2.transform(train['groupId'])

Le3=LabelEncoder()
Le3.fit(train['matchId'])
train['Encoded_matchId']=Le3.transform(train['matchId'])

Le4=LabelEncoder()
Le4.fit(train['matchType'])
train['Encoded_matchType']=Le4.transform(train['matchType'])
```

```python
print(train.head(5))
```

# **Splitting Data into Train set and Test set**

```python
y=train['winPlacePerc']
X=train[['Encoded_Id','Encoded_groupId','Encoded_matchId','assists','boosts','damageDealt','DBNOs','headshotKills','heals','killPlace','killPoints','kills','killStreaks','longestKill','matchDuration','Encoded_matchType','maxPlace','numGroups','rankPoints','revives','rideDistance','roadKills','swimDistance','teamKills','vehicleDestroys','walkDistance','weaponsAcquired','winPoints']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=7)
```

# **Predition using Random Forest Regressor**

```python
rf=RandomForestRegressor()
rf.fit(X_train,y_train)
```

# **Score**

```python
predictions_rf=rf.predict(X_test)
Accuracy = rf.score(X_test,y_test)
print(Accuracy)
```
