```python
df_train = df_train.drop(columns=cols_to_be_removed,axis = 1)
df_test = df_test.drop(columns=cols_to_be_removed,axis = 1)
```

```python
def one_hot_encoding(df,col):
    one_hot_encoder=ce.OneHotEncoder(cols=col,return_df=True,use_cat_names=True)
    df_final = one_hot_encoder.fit_transform(df)
    return df_final
```

```python
df_train = one_hot_encoding(df_train,"matchType")
```

```python
df_test = one_hot_encoding(df_test,"matchType")
```

```python
X_train = df_train.drop("winPlacePerc",axis = 1)
X_test = df_test.copy()
Y_train = df_train["winPlacePerc"]
```

```python
X_train = normalize(X_train)
X_test = normalize(X_test)
```

```python
gradient_boosting_regr = GradientBoostingRegressor()
gradient_boosting_regr.fit(X_train, Y_train)
```

```python
df_train.to_csv("df_train_cleaned.csv",index = False)
df_test.to_csv("df_test_cleaned.csv",index = False)
```

```python
Y_pred = gradient_boosting_regr.predict(X_test)
```

```python
submission = pd.DataFrame()
submission["id"] = df_test_copy["Id"]
submission["winPlacePerc"] = Y_pred
```
