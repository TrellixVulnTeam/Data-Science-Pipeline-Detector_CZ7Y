- weaponsAcquired 0.6 - Number of weapons picked up. 捡了多少武器
- boosts 0.6 - Number of boost items used. 回血物品使用数量

## low relate features
- killPlace -0.7 - Ranking in match of number of enemy players killed. 极高的负相关，这里是杀敌数？还是敌人杀人数的排名？？ 杀敌数量排名越高，名次越低？？？

## normal relate features
- damageDealt 0.4 - Total damage dealt. Note: Self inflicted damage is subtracted. 总伤害
- heals 0.4 - Number of healing items used. 加血总次数
- kills 0.4 - Number of enemy players killed. 杀敌次数？？
- killStreaks 0.4 - Max number of enemy players killed in a short amount of time. 近距离 杀敌数
- longstkill 0.4 - Longest distance between player and player killed at time of death. This may be misleading, as downing a player and driving away may lead to a large longestKill stat. 远距离 杀敌数， 可能不准

- assists 0.3 - Number of enemy players this player damaged that were killed by teammates. 助攻，队友做最后击杀
- DBNOs 0.3 - Number of enemy players knocked. 击倒杀
- headshotKills 0.3 - Number of enemy players killed with headshots.  爆头数
- rideDistance 0.3 - Total distance traveled in vehicles measured in meters. 车载行驶距离

## none relate features
- killPoints - Kills-based external ranking of player. (Think of this as an Elo ranking where only kills matter.) If there is a value other than -1 in rankPoints, then any 0 in killPoints should be treated as a “None”. 杀人排名
- matchDuration - Duration of match in seconds. 比赛时长
- maxPlace - Worst placement we have data for in the match. This may not match with numGroups, as sometimes the data skips over placements. 最差的队伍数量， 确实没什么关系
- numGroups - Number of groups we have data for in the match. 有多少队人，这个确实弱关系
- rankPoints - Elo-like ranking of player. This ranking is inconsistent and is being deprecated in the API’s next version, so use with caution. Value of -1 takes place of “None”.  淘汰的字段，确实无关
- roadKills - Number of kills while in a vehicle. 在车上杀人，这个和开车时长一样貌似可以说明他不是机器人
- teamKills - Number of times this player killed a teammate. 杀队友的次数？？？ 这个什么鬼。。。
- winPoints - Win-based external ranking of player. (Think of this as an Elo ranking where only winning matters.) If there is a value other than -1 in rankPoints, then any 0 in winPoints should be treated as a “None”.
