```python
train = pd.read_csv("../input/train_V2.csv")
test = pd.read_csv("../input/test_V2.csv")
```

```python
train.columns
```

```python
train.head()
```

```python
feature = ["boosts","heals","kills","longestKill","matchDuration","revives","rideDistance","weaponsAcquired","winPoints"]
label = "winPlacePerc"
```

```python
X_train = train[feature]
```

```python
Y = train[label]
```

```python
X_train.isnull().sum()
```

```python
Y.isnull().sum()
```

```python
Y = Y.fillna(0)
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,Y)
```

```python
lm.score(X_train,Y)
```

```python
X_test = test[feature]
```

```python
lm.predict(X_test)
```

```python
test[label] = lm.predict(X_test)
```
