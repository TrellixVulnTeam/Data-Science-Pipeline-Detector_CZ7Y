# **데이터 불러오기 (data load)**

**python은 자동으로 자료형을 할당하는 경우가 많습니다.**

**미리 자료형을 지정해준다면 많은 메모리를 절약할 수 있습니다.**

**983.9MB의 데이터가 274.1MB로 절약된 것을 볼 수 있습니다.**

**python often assigns data types automatically.**

**If you specify the data type in advance, you can save a lot of memory.**

**983.9MB of data was saved to 274.1MB.**

```python
train = pd.read_csv("../input/pubg-finish-placement-prediction/train_V2.csv")
test = pd.read_csv("../input/pubg-finish-placement-prediction/test_V2.csv")
train.info()
```

```python
dtypes = {
        'assists'           : 'uint8',
        'boosts'            : 'uint8',
        'damageDealt'       : 'float16',
        'DBNOs'             : 'uint8',
        'headshotKills'     : 'uint8', 
        'heals'             : 'uint8',    
        'killPlace'         : 'uint8',    
        'killPoints'        : 'uint16',    
        'kills'             : 'uint8',    
        'killStreaks'       : 'uint8',    
        'longestKill'       : 'float16',
        'matchDuration'     : 'uint8',
        'maxPlace'          : 'uint8',    
        'numGroups'         : 'uint8',    
        'rankPoints'        : 'uint8',
        'revives'           : 'uint8',    
        'rideDistance'      : 'float16',    
        'roadKills'         : 'uint8',    
        'swimDistance'      : 'float16',    
        'teamKills'         : 'uint8',    
        'vehicleDestroys'   : 'uint8',    
        'walkDistance'      : 'float16',    
        'weaponsAcquired'   : 'uint8',    
        'winPoints'         : 'uint8', 
        'winPlacePerc'      : 'float16' 
}
```

**pandas의 read_csv 기능에 파일 경로를 입력하면 csv 파일을 읽을 수 있습니다**

**경로는 copy file path 기능을 이용하면 편합니다.(data 파일 오른쪽에 copy 버튼이 있습니다.)**

**You can read the csv file by entering the file path in the read_csv function of Pandas.**

**It is convenient to use the copy file path function. (There is a copy button to the right of the data file.)**

```python
train = pd.read_csv("../input/pubg-finish-placement-prediction/train_V2.csv",dtype=dtypes)
test = pd.read_csv("../input/pubg-finish-placement-prediction/test_V2.csv",dtype=dtypes)
train.info()
```

**고윳값이 많은 id들과 matchtype는 drop하였습니다.**

**The match type and id with many unique value were dropped.**

```python
train = train.dropna()
train = train.drop(["Id","groupId","matchId","matchType"],axis=1)
test = test.drop(["Id","groupId","matchId","matchType"],axis=1)
```

**예측에 사용할 열과 예측할 열을 나눕니다.**

**Divide the column to be used for prediction and the column to be predicted.**

```python
X = train.drop("winPlacePerc",axis=1)
y = train["winPlacePerc"]
```

# **학습 및 제출 (training and submission)**

**LGBMRegressor를 이용하여 학습데이터를 학습시키고 test 데이터를 예측해봅시다.**

**Let's train train data using LGBMRegressor and predict test data.**

```python
model = LGBMRegressor()
model.fit(X,y)
pred = model.predict(test)
```

**예측한 값을 csv로 제출해봅시다**

**Let's submit the predicted value to csv.**

```python
sample_submission = pd.read_csv("../input/pubg-finish-placement-prediction/sample_submission_V2.csv",index_col = "Id")
sample_submission["winPlacePerc"] = pred
sample_submission.to_csv("submission.csv")
```
