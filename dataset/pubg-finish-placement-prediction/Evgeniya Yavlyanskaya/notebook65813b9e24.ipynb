```python
import pandas as pd
from pandas import Series,DataFrame
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train_data = pd.read_csv("/kaggle/input/pubg-finish-placement-prediction/train_V2.csv")
```

```python
train_data.head()
```

```python
plt.title('Плотность распределения количества убийств игроком для тренировочной выборки')
_ = sns.distplot(train_data.kills)
```

```python
plt.title('Плотность распределения количества урона для тех, кто совершил 0 убийств')
_ = sns.distplot(train_data[train_data.kills == 0].damageDealt)
```

```python
plt.title('Плотность распределения пройденной дистанции')
_ = sns.distplot(train_data.walkDistance)
```

```python
train_data.loc[train_data['kills'] == 1, 'kills_n'] = 'kills_1'
train_data.loc[(train_data['kills'] >= 2) & (train_data['kills'] <= 3), 'kills_n'] = 'kills_23'
train_data.loc[(train_data['kills'] >= 4) & (train_data['kills'] <= 5), 'kills_n'] = 'kills_45'
train_data.loc[train_data['kills'] > 5, 'kills_n'] = 'kills_5+'
```

```python
sns.boxplot(x='kills_n', y='winPoints', data=train_data)
```

```python
plt.title('Боксплот персентили победы от количества использованных heals/boosts')
_ = sns.boxplot(data = train_data, x = 'heals', y = 'winPlacePerc')
```

```python
plt.title('Боксплот персентили победы от количества использованных heals/boosts')
_ = sns.boxplot(data = train_data, x = 'boosts', y = 'winPlacePerc')
```
