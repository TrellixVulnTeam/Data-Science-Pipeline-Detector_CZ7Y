<h1 style='background:brown; border:0; color:white'><center>Import Packages<center><h1>

```python
import sys
sys.path.append('../input/xlrd-python') #required for autoviz
from autoviz.AutoViz_Class import AutoViz_Class

AV = AutoViz_Class();
```

<a id="1"></a>
<h1 style='background:brown; border:0; color:white'><center>Visualize Training Data<center><h1>

```python
filename = "../input/tabular-playground-series-mar-2021/train.csv"
sep = ","
dft = AV.AutoViz(
    filename,
    sep=",",
    depVar="target",
    dfte=None,
    header=0,
    verbose=1,
    lowess=False,
    chart_format="svg",
    max_rows_analyzed=300000,
    max_cols_analyzed=32,
)
```

<a id="1"></a>
<h1 style='background:brown; border:0; color:white'><center>Visualize Testing Data<center><h1>

```python
filename = "../input/tabular-playground-series-mar-2021/test.csv"
sep = ","
dft = AV.AutoViz(
    filename,
    sep=",",
    depVar="", # No dependant variable
    dfte=None,
    header=0,
    verbose=1,
    lowess=False,
    chart_format="svg",
    max_rows_analyzed=300000,
    max_cols_analyzed=32,
)
```
