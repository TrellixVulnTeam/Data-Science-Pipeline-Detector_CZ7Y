```python
x=train_df.iloc[:,:-1]
y=train_df.iloc[:,-1]
print("class label: ",np.unique(y))
```

```python
#spration different types of columns
cat_features = [feature for feature in x.columns if 'cat' in feature]
cont_features = [feature for feature in x.columns if 'cont' in feature]
#spration different types of columns
cat_features_test = [feature for feature in test_df.columns if 'cat' in feature]
cont_features_test = [feature for feature in test_df.columns if 'cont' in feature]
```

```python
#trian_data
categorical_features = pd.DataFrame(x[cat_features])
numric_train=pd.DataFrame(x[cont_features])
#test_data
catgo_features_test = pd.DataFrame(test_df[cat_features_test])
numric_test=pd.DataFrame(test_df[cont_features_test])
```

```python
# importing one hot encoder 
from sklearn.preprocessing import OneHotEncoder

# creating one hot encoder object 
onehotencoder = OneHotEncoder(handle_unknown='ignore', sparse=False)
 
OHE_cat_train = onehotencoder.fit_transform(categorical_features)
OHE_cat_test = onehotencoder.fit_transform(catgo_features_test)

#Tranforming OHE_cat_train from array to dataframe
OHE_cat_train = pd.DataFrame(OHE_cat_train)
OHE_cat_test = pd.DataFrame(OHE_cat_test)
cont_features=pd.DataFrame(cont_features)
cont_features_test=pd.DataFrame(cont_features_test)
```

```python
#Concatenating numerical with categorical 
df_train = pd.concat([numric_train, OHE_cat_train], axis=1)
df_test = pd.concat([numric_test, OHE_cat_test], axis=1)
#printing to verify 
```

```python
from sklearn.utils.class_weight import compute_class_weight
 
classes = np.unique(y)
weights = compute_class_weight(class_weight='balanced', classes=classes, y=y)
class_weights1 = dict(zip(classes, weights))
#
clf = CatBoostClassifier(loss_function='Logloss', random_state=42, verbose=False,class_weights=class_weights1)
#clf.fit(df_train, y)
```

```python
cbc_score = cross_val_score(clf,df_train, y, scoring='roc_auc', cv=5)
print('cbc_score: ', cbc_score.mean())
```
