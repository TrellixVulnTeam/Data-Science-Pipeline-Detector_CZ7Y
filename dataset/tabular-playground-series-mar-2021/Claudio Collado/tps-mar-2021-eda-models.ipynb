## 1.2 Continuous variables

```python
def grafico_distribucion(columna,dataset):
    
    import random
    
    datos = dataset[columna]
    
    paleta = sns.color_palette()
    color = random.choice(paleta)

    sns.displot(datos, color=color)
    
    plt.show()
```

```python
#Distribution for continuous variables - Train

for columna in cont_columns:
    grafico_distribucion(columna,df_train)
```

```python
#Distribution for continuous variables - Test

for columna in cont_columns:
    grafico_distribucion(columna,df_test)
```

# 2. Encoding for categorical variables

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
OH_encoder = OneHotEncoder(sparse=False)
```

```python
#Categorical column encoding

#OH_cols_train = pd.DataFrame(OH_encoder.fit_transform(df_train[cat_columns]))
#OH_cols_test = pd.DataFrame(OH_encoder.transform(df_test[cat_columns]))
```

As seen from the previous error that {'BU', 'BW', 'CA', 'DG', 'EJ', 'JM', 'KE', 'KM'} in cat10 column are present in test data but not present in training data

In principle I will not use this column for train and test

```python
cat_columns_reduce = ['cat0','cat1','cat2','cat3','cat4','cat5','cat6','cat7','cat8','cat9','cat11','cat12','cat13','cat14','cat15','cat16','cat17','cat18']
```

```python
#Categorical column encoding

OH_cols_train = pd.DataFrame(OH_encoder.fit_transform(df_train[cat_columns_reduce]))
OH_cols_test = pd.DataFrame(OH_encoder.transform(df_test[cat_columns_reduce]))
```

```python
#Column name for categorical column - Train

OH_cols_train.columns = OH_encoder.get_feature_names(['cat0', 'cat1', 'cat2', 'cat3', 'cat4', 'cat5', 'cat6', 'cat7','cat8','cat9','cat11','cat12','cat13','cat14','cat15','cat16','cat17','cat18'])
```

```python
#Column name for categorical column - Test

OH_cols_test.columns = OH_encoder.get_feature_names(['cat0', 'cat1', 'cat2', 'cat3', 'cat4', 'cat5', 'cat6', 'cat7','cat8','cat9','cat11','cat12','cat13','cat14','cat15','cat16','cat17','cat18'])
```

```python
OH_cols_train.head()
```

```python
OH_cols_test.head()
```

```python
#Remove categorical columns without encoding and add new ones with encoding - Train

df_train_OH = df_train.drop(df_train[cat_columns],axis=1)
df_train_OH = pd.concat([df_train_OH,OH_cols_train],axis=1)
df_train_OH.head()
```

```python
df_train_OH.columns
```

```python
len(df_train_OH.columns)
```

```python
#Remove categorical columns without encoding and add new ones with encoding - Test

df_test_OH = df_test.drop(df_test[cat_columns],axis=1)
df_test_OH = pd.concat([df_test_OH,OH_cols_test],axis=1)
df_test_OH.head()
```

```python
df_test_OH.columns
```

```python
len(df_test_OH.columns)
```

# 3. Datasets for Model training

```python
#Target for training

target = df_train['target']
target.head()
```

```python
#Train dataset

df_train_OH = df_train_OH.drop(columns=['target'])
df_train_OH.head()
```

```python
len(df_train_OH.columns)
```

# 5. LightGBM Model


```python
from lightgbm import LGBMRegressor
```

```python
hyperparameters = {"random_state": 42,
                   "boosting_type": "gbdt",
                   "objective": "binary",
                   "metric": "auc",
                   "colsample_bytree": 0.25,
                   "learning_rate": 0.01,
                   "max_depth": 54,
                   "num_leaves": 100,
                   "min_child_samples": 275,
                   "n_estimators": 2300,
                   "cat_smooth": 40.0,
                   "max_bin": 512,
                   "min_data_per_group": 100,
                   "bagging_freq": 1,
                   "bagging_fraction": 0.80,
                   "cat_l2": 12.0}
```

```python
model = LGBMRegressor(**hyperparameters)
```

```python
model.fit(df_train_OH, target)
```

# 6. Prediction and submit

```python
sample_submission['target'] = model.predict(df_test_OH)

sample_submission.to_csv('lightgbm.csv',index=False)
```
