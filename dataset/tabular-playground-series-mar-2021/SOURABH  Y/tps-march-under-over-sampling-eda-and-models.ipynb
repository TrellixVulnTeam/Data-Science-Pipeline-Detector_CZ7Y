```python
import matplotlib.pyplot as plt

def pie_chart(col):
    
    rating_data = data.groupby(col)[['id']].count()

    plt.figure(figsize = (12, 8))
    plt.pie(rating_data['id'], 
            autopct = '%1.000f%%', 
            startangle = 140, 
            pctdistance = 1.1, 
            shadow = True)
    
    plt.axis('equal')

    cols = []
    for index, row in rating_data.iterrows():
        cols.append(index)
    
    plt.legend(cols)
    
    plt.title('Pie Chart by ' +str(col), loc='center')
    
    plt.show()
```

```python
col=['id', 'cat0', 'cat1', 'cat2', 'cat3', 'cat4', 'cat5', 'cat6', 'cat7',
       'cat8', 'cat9', 'cat10', 'cat11', 'cat12', 'cat13', 'cat14', 'cat15',
       'cat16', 'cat17', 'cat18', 'cont0', 'cont1', 'cont2', 'cont3', 'cont4',
       'cont5', 'cont6', 'cont7', 'cont8', 'cont9', 'cont10']
X=data[col]
Y=data['target']
```

```python
import keras 
from sklearn.model_selection import train_test_split 
from keras.utils import to_categorical
from keras.preprocessing.text import Tokenizer
from keras.models import Sequential
from keras.layers import Dense, Activation, Embedding, LSTM,Dropout
```

```python
def create_model():
        model = Sequential()
        model.add(Dense(31,input_dim=31,activation='relu'))
        model.add(Dense(62,activation='relu'))
        model.add(Dropout(0.4))
        model.add(Dense(62,activation='relu'))
        model.add(Dropout(0.4))
        model.add(Dense(30,activation='relu'))
        model.add(Dropout(0.4))
        model.add(Dense(12,activation='relu'))
        model.add(Dense(1, activation='sigmoid'))
        model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
        return model
```

```python
model=create_model()
model.summary()
```

```python
train_x,test_x,train_y,test_y=train_test_split(X,Y,test_size=0.15)
```

```python
hist=model.fit(train_x,train_y,epochs=10,batch_size=140)
```

```python
model.evaluate(test_x,test_y)
```

```python
y_pred=model.predict(test_x)
```

```python
def give_app(y_pred):
    pred=[]
    for i in y_pred:
        if i>0.500000001:
            pred.append(1)
        else:
            pred.append(0)
    return np.asarray(pred)
            
```

```python
pred=give_app(y_pred)
```

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score(test_y, y_pred)
```
