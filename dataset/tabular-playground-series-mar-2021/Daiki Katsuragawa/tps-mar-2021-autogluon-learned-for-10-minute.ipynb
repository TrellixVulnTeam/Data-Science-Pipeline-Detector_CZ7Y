```python
import pandas as pd
from autogluon.tabular import TabularDataset, TabularPredictor

train= TabularDataset('../input/tabular-playground-series-mar-2021/train.csv')
test = TabularDataset('../input/tabular-playground-series-mar-2021/test.csv')

label = 'target'

# If the value of time_limit is too small, we will get the following error:
# ValueError: AutoGluon did not successfully train any models
time_limit = 60 * 10

predictor = TabularPredictor(
    label=label,
    eval_metric='roc_auc',
    learner_kwargs={'ignored_columns': ['id']},
).fit(
    train,
    presets='best_quality',
    time_limit=time_limit,
)

submission = pd.read_csv('../input/tabular-playground-series-mar-2021/sample_submission.csv')
submission[label] = predictor.predict_proba(test, as_multiclass=False)
submission.to_csv('submission.csv', index=False)
submission.head()
```
