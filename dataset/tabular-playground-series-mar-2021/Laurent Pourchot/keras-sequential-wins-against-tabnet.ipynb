```python
X=pd.read_csv('../input/keras-denoiser/1X.csv')
y=pd.read_csv('../input/marscompetition/1y.csv')
```

```python
X_train = X.to_numpy()
y_train = y['target'].to_numpy().squeeze()
X_train,X_valid,y_train,y_valid=train_test_split(X_train,y_train,random_state=41)
```

<h1>TabNet training <h1>

```python
classifier = TabNetClassifier(verbose=1,seed=42)
classifier.fit(X_train=X_train, y_train=y_train,
               eval_set=[(X_train, y_train), (X_valid, y_valid)],
               patience=5,max_epochs=100,
               eval_metric=['auc'])
```

<h1> Keras Sequential <h1>

```python
model = Sequential()
model.add(Dense(300, activation='relu',input_dim=30))
model.add(Dense(30, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

```python
tf.keras.metrics.AUC(
    num_thresholds=200,
    curve="ROC",
    summation_method="interpolation",
    name=None,
    dtype=None,
    thresholds=None,
    multi_label=False,
    label_weights=None,
)
tf.keras.optimizers.SGD(learning_rate=0.0001, name='SGD')
callback = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=5)
model.compile(optimizer='SGD', loss='binary_crossentropy', metrics=[tf.keras.metrics.AUC()])
```

```python
history=model.fit(x=X_train,
                  y=y_train,
                  validation_data=(X_valid, y_valid),
                  batch_size=256,
                  epochs=100,
                  shuffle=False,
                  verbose=2)
```
