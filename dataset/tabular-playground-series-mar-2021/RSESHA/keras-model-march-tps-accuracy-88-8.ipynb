# deep_autoviml is a tensorflow 2.5-powered, keras-based, model and pipeline building auto-ml library
## It will automatically do the following:
- Load a preselected keras model template that will select the hidden layers
- Load train, test files containing training and testing data
- For NLP task: select a BERT or USE model along with preprocessors
- Run a tuner named Storm-Tuner to select the best hyper params
- Train and run predictions using just one line of code each
## Keras preprocessing layers enable you to encapsulate feature engineering and hidden layers into the model itself. This makes the process for training and predictions the same: just feed input data (in the form of files or dataframes) and the model will take care of all preprocessing before predictions.

```python
from load_kaggle import load_kaggle
```

```python
subm, train, test = load_kaggle()
print(train.shape)
train.head()
```

```python
from deep_autoviml import deep_autoviml as deepauto
```

```python
######   D E F A U L T S    S E T T I N G S   F O R   D E E P    A U T O  V I M L ###
keras_model_type =  "fast1" ## try "fast" first, then "fast1" and "fast2", finally "auto"
### always set early_stopping to True first and then change it to False
#### You always need 15 max_trials to get something decent #####
#### always set tuner to "storm" and then "optuna". 
### NLP char limit kicks off NLP processing. Feature Cross later.
model_options = {'nlp_char_limit':50, 'cat_feat_cross_flag':"both",
                 'max_trials': 10, "tuner": "storm"}
keras_options = {"patience":10, 'class_weight': True, 'early_stopping': True, 
                 'lr_scheduler': ''}
project_name = 'TPS_March'
save_model_flag = False
```

```python
target = 'target'
```

```python
model, cat_vocab_dict = deepauto.fit(train, target, keras_model_type=keras_model_type,
		project_name=project_name, keras_options=keras_options,  
		model_options=model_options, save_model_flag=True, use_my_model='',
		model_use_case='', verbose=1)
```

```python
predictions = deepauto.predict(model, project_name, test_dataset=test,
                                 keras_model_type=keras_model_type, 
                                 cat_vocab_dict=cat_vocab_dict)
```

```python
y_preds = predictions[0][:,1]
y_preds
```

```python

subm[target] = y_preds
subm.head()
```

```python
subm.to_csv('submission.csv',index=False)
subm.head()
```
