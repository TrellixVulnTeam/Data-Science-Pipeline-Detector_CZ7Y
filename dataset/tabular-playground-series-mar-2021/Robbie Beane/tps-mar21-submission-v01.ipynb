# Tabular Playground Series - March 2021

```python
import pandas as pd
import joblib
```

# Load Data

```python
test = pd.read_csv('../input/tabular-playground-series-mar-2021/test.csv')

test.drop('id', axis=1, inplace=True)

test.head()
```

# Load Model

```python
preprocessor = joblib.load('../input/tps-mar21-models/tps_preprocessor_01.joblib')
model = joblib.load('../input/tps-mar21-models/tps_model_01.joblib')
print(type(model))
```

# Preprocessing

```python
X_test = preprocessor.transform(test)
print(X_test.shape)
```

# Test Predictions

```python
test_pred = model.predict_proba(X_test)
print(test_pred.shape)
```

```python
print(test_pred[:10, :])
```

```python
submission = pd.read_csv('../input/tabular-playground-series-mar-2021/sample_submission.csv')
submission.head()
```

```python
submission.target = test_pred[:,1]
submission.head()
```

```python
submission.to_csv('submission.csv', index=False, header=True)
```
