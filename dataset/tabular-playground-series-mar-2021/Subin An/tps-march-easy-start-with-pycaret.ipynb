```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

train = pd.read_csv('/kaggle/input/tabular-playground-series-mar-2021/train.csv')
test = pd.read_csv('/kaggle/input/tabular-playground-series-mar-2021/test.csv')
sample = pd.read_csv('/kaggle/input/tabular-playground-series-mar-2021/sample_submission.csv')

train.describe()
```

```python
!pip install pycaret
from pycaret.classification import *
```

```python
setup(data = train, 
      target = 'target',
      numeric_imputation = 'mean',
      fold=5,
      silent = True)
```

```python
model_name = 'lightgbm'
model = create_model(model_name)
```

```python
tuned_model = tune_model(model)
```

```python
plot_model(estimator = tuned_model, plot = 'learning')
```

```python
plot_model(estimator = tuned_model, plot = 'feature')
```

```python
predictions = predict_model(tuned_model, data = test)
sample['target'] = predictions['Label']
sample.to_csv(f'submission_{model_name}.csv',index=False)
sample.head()
```
