```python
df_train = pd.read_csv('../input/tabular-playground-series-mar-2021/train.csv')
df_test = pd.read_csv('../input/tabular-playground-series-mar-2021/test.csv')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df_train.columns
```

```python
df_train_enc = df_train[['cat0', 'cat1', 'cat2', 'cat3', 'cat4', 'cat5', 'cat6', 'cat7',
       'cat8', 'cat9', 'cat10', 'cat11', 'cat12', 'cat13', 'cat14', 'cat15',
       'cat16', 'cat17', 'cat18']]
```

```python
df_train_cont = df_train[['cont0', 'cont1', 'cont2', 'cont3', 'cont4',
       'cont5', 'cont6', 'cont7', 'cont8', 'cont9', 'cont10', 'target']]
```

```python
df_train_encoded = df_train_enc.apply(LabelEncoder().fit_transform)
```

```python
df_train_encoded.head()
```

```python
df_train = pd.concat([df_train_encoded,df_train_cont],axis=1)
```

```python
df_test_enc = df_test_cat.apply(LabelEncoder().fit_transform)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
position=1
plt.figure(figsize=(15,45))
for column in ['cont0', 'cont1', 'cont2', 'cont3', 'cont4',
       'cont5', 'cont6', 'cont7', 'cont8', 'cont9', 'cont10','target']:
    plt.subplot(9,2,position)
    if column in ['target']:
        sns.countplot(df_train[column])
    else:
        sns.distplot(df_train[column])
    plt.title(column, fontsize=20)
    position+=1
plt.show()
```
