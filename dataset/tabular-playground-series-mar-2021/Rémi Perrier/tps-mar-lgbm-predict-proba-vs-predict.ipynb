## Just a small notebook to show you that you should use predict_proba for this competition

```python
# Libraries
import pandas as pd
from lightgbm import LGBMClassifier
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import train_test_split
```

```python
# Read data
train = pd.read_csv('../input/tabular-playground-series-mar-2021/train.csv', index_col=0)

# Predictors & target
predictors = train.columns[:-1]
target = train.columns[-1]

# Data processing
cat_cols = [col for col in predictors if 'cat' in col]
train[cat_cols] = train[cat_cols].astype('category')
train[target] = train[target].astype('category')

# Split
X_train, X_test, y_train, y_test = train_test_split(train[predictors], train[target])
```

## Train model

```python
# Create and train model
model = LGBMClassifier()
model.fit(X_train, y_train)
```

## Using predict

```python
# Predict
model.predict(X_test)
```

Array of 'class label' given to each observation.

```python
# Score
y_pred = model.predict(X_test)
roc_auc_score(y_test, y_pred)
```

## predict_proba

```python
# Predict
model.predict_proba(X_test)
```

Array of 'probability to be in each class' given to each observation.  
We need to use the probability to be in the class '1'.  

```python
# Score
y_pred = model.predict_proba(X_test)[:, 1]
roc_auc_score(y_test, y_pred)
```

## With *predict_proba[X_test][:, 1]* the score is far better despite using the exact same model !!
