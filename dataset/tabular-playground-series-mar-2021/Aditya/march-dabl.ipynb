```python
#import train test files
train = pd.read_csv('/kaggle/input/tabular-playground-series-mar-2021/train.csv')
test = pd.read_csv('/kaggle/input/tabular-playground-series-mar-2021/test.csv')
sample_submission = pd.read_csv('/kaggle/input/tabular-playground-series-mar-2021/sample_submission.csv')
```

```python
types = dabl.detect_types(train)
print(types) 
```

## Exploratory Data analysis with dabl

```python
dabl.plot(train,target_col='target')
```

```python
ec = dabl.SimpleClassifier(random_state=0).fit(train, target_col="target") 
```

## Model Building with dabl

```python
dabl.explain(ec) 
```

```python
pred = ec.predict(test)
```

```python
sample_submission['target'] = ec.predict_proba(test)[:, 1]
sample_submission.to_csv('submission.csv',index=False)
```
