```python
import numpy as np
import pandas as pd
import sweetviz as sv
import category_encoders as ce
```

```python
train_data = pd.read_csv("../input/tabular-playground-series-mar-2021/train.csv")

train_data
```

```python
categorical_features = train_data.columns[1:20]

categorical_features
```

```python
for cat in categorical_features:
    print("Cat\t\tLen\t\tValues")
    print(f"{cat}\t\t{len(train_data[cat].unique())}\t\t{train_data[cat].unique()}")
    print()
```

Vamos transformar as categorias categoricas usando o mesmo tranformer do CatBoost.

```python
cat_enconder = ce.cat_boost.CatBoostEncoder(verbose = 1, cols = categorical_features)

cat_enconder
```

```python
train_data_enconded = cat_enconder.fit_transform(train_data[train_data.columns[1:-1]], train_data["target"])

train_data_enconded
```

Gera um relatório com o sweetviz para verificar a distribuição das categorias que foram "encodadas".

```python
train_data_enconded = train_data_enconded.merge(train_data["target"], left_index = True, right_index = True)

train_data_enconded
```

```python
report = sv.analyze(train_data_enconded, "target")

report.show_html("report.html", False)
```

# Transforma o test

Vamos aplicar as mesmas transformações usadas no conjunto de treino, para o conjunto de test.

```python
test_data = pd.read_csv("../input/tabular-playground-series-mar-2021/test.csv")

test_data
```

```python
test_data_enconded = cat_enconder.transform(test_data[test_data.columns[1:]])

test_data_enconded
```

```python
test_data_enconded = test_data_enconded.merge(test_data["id"], left_index = True, right_index = True)

test_data_enconded
```

# Salva os arquivos

```python
train_data_enconded.to_csv("train_enconded.csv", index = True)
```

```python
test_data_enconded.to_csv("test_enconded.csv", index = True)
```

# Gera um relatório que compara o train e test

```python
report = sv.compare([train_data_enconded, "Train"], [test_data_enconded, "Test"], "target")

report.show_html("report2.html", False)
```
