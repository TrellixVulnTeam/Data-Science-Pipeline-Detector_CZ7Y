## Importa as bibliotecas

```python
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt

from pandas_profiling import ProfileReport

from catboost import CatBoostClassifier

from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold

from sklearn.metrics import roc_auc_score
```

## Lê os dados

```python
data_train = pd.read_csv("../input/tabular-playground-series-mar-2021/train.csv")

data_train.head(5)
```

## Modelo para teste

```python
 X_train, X_test, y_train, y_test = train_test_split(data_train[data_train.columns[1:-1]], data_train["target"], test_size=0.2, random_state=42)
```

```python
model = CatBoostClassifier(
    eval_metric="AUC",
    verbose=True,
    cat_features = data_train.columns[1:20]
)
```

```python
model.fit(X_train, y_train)
```

```python
preds_class = model.predict(X_test)

roc_auc_score(y_test, preds_class)
```
