```python
import catboost
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import KFold
from sklearn.metrics import roc_auc_score
```

## Lendo os dados

```python
train_df = pd.read_csv("../input/tabular-playground-series-mar-2021/train.csv")

train_df
```

```python
train_df.info()
```

## Testando as ideias

### Feature importance do catboost

```python
catboost_params = {
    "depth": [2, 3, 4, 8, 10],
    "iterations": [250, 500, 1000],
    "random_seed": [1],
    "learning_rate": [0.1, 0.5]
}
cv_dataset = catboost.Pool(data = train_df.drop(["target", "id"], axis = 1), label = train_df["target"], cat_features = train_df.columns[1:20])

model = catboost.CatBoostClassifier(
    eval_metric = "AUC",
    loss_function = "Logloss",
    task_type = "GPU"
)

best_params, cv_results = model.grid_search(catboost_params, cv_dataset, cv = 2, stratified = True, plot = True, verbose = True)
```
