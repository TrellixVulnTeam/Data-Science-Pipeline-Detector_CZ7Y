# Blend Boosting study on Tabular Playground Series - Mar 2021:
Here I share with you a systematic blend boosting study on Tabular Playground Series - Mar 2021. (https://www.kaggle.com/c/tabular-playground-series-mar-2021). I just collect some submission files on the kaggle.

Basically, I start to analysis of correlations, then decide to sort them according to their sum of correlation values in between. This lets me divide ~15 scores into 4 subgroups. Then I make internal linear calibration in each subgroup by considering their scores on the Kaggle. If you spend much more time, you can always achieve much betters scores.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# loading dummy submission file
sub_file = pd.read_csv(r'../input/tspmar2021/submission_blend_1.csv')

# loading data including 15 best scores
df_sub = pd.read_csv(r'../input/tspmar2021/best_blend_1.csv')
df_sub = df_sub.iloc[:, :15]

# a rough correlation based visualization of 15 best scores
plt.figure(figsize=(10,10))
sns.heatmap(df_sub.corr(), cmap='Spectral')
plt.ylabel('file index numbers')
plt.xlabel('file index numbers')
plt.show()
```

```python
# basic analysis and visualization of subgroups in different color.
plt.figure(figsize=(12, 5))
df_mean_corr = pd.DataFrame({'mean_corr': df_sub.corr().mean()})
df_mean_corr = df_mean_corr.sort_values('mean_corr', ascending=False)
df_mean_corr = df_mean_corr.reset_index()

plt.plot(df_mean_corr.index[:7], df_mean_corr['mean_corr'].values[:7], 'o', ms=10)
plt.plot(df_mean_corr.index[7:10], df_mean_corr['mean_corr'].values[7:10], 'o', ms=10)
plt.plot(df_mean_corr.index[10:11], df_mean_corr['mean_corr'].values[10:11], 'o', ms=10)
plt.plot(df_mean_corr.index[11:12], df_mean_corr['mean_corr'].values[11:12], 'o', ms=10)
plt.plot(df_mean_corr.index[12:13], df_mean_corr['mean_corr'].values[12:13], 'o', ms=10)
plt.plot(df_mean_corr.index[13:14], df_mean_corr['mean_corr'].values[13:14], 'o', ms=10)
plt.plot(df_mean_corr.index[14:], df_mean_corr['mean_corr'].values[14:], 'o', ms=10)
plt.xticks([*range(len(df_mean_corr))], df_mean_corr['index'].tolist())
plt.title('determination of sub_groups')
plt.ylabel('a correlation related index')
plt.xlabel('file index numbers')
plt.show()
```

```python
# a linear combination to achieve much better scores
df_sub['weighted_avg'] = abs(1 * (
    -15 * ( 5 * df_sub['6'] +  5 * df_sub['7'] +  5 * df_sub['8'] + 4 * df_sub['9']
         + 3 * df_sub['10'] + 2 * df_sub['12'] + 1 * df_sub['14']) / 25 +
    15 * (80 *  df_sub['2'] - 43 * df_sub['3'] - 20 * df_sub['11']) / 17 +
   150 * (150 * df_sub['0'] + 50 * df_sub['1'] +  20 * df_sub['4'] + 10 * df_sub['5'] + 10 * df_sub['13']) / 240
) / 150 )

# # create the final submission file
submission = pd.DataFrame({'ID': sub_file.ID, 'target': df_sub['weighted_avg'].tolist()})
submission.to_csv(r'submission.csv', index=False)
```
