# Tabular Playground March EDA

    

# Import Library and Loading Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
import warnings
warnings.filterwarnings("ignore")
```

```python
train_dataset = pd.read_csv("../input/tabular-playground-series-mar-2021/train.csv")
train_dataset.drop(labels=['id'], axis=1, inplace=True)
train_dataset.head()
```

```python
print("Shape of train_dataset: ",train_dataset.shape)
```

```python
train_dataset.info()
```

# EDA (Evaluation Data Analysis) ðŸ“Š

## Distplot

```python
fig, ax = plt.subplots(nrows=4, ncols=3, figsize=(10, 20))

for i, col in enumerate(train_dataset.iloc[:, 19:30]):
    x = i // 3
    y = i%3
    sns.distplot(train_dataset[col], ax=ax[x, y], color='blue')
    
fig.delaxes(ax[3, 2])
plt.tight_layout()
plt.show()
```

## Encoding

```python
def encoder(dataset):
    le = LabelEncoder()
    for i in range(19):
        dataset.iloc[:, i] = le.fit_transform(dataset.iloc[:, i])
    return dataset

train_dataset = encoder(train_dataset)
train_dataset.head()
```

```python
print(train_dataset.iloc[:, :15].describe())
print(train_dataset.iloc[:, 15:31].describe())
```

## Correlation Matrix

```python
corr_matrix = train_dataset.corr()
corr_matrix['target'].sort_values()
```

```python
corr_matrix['target'].sort_values().plot(kind="bar")
```

```python
sns.clustermap(corr_matrix, annot=True, fmt=".2f", figsize=(25,25))
plt.title("Correlation Between Features")
plt.show()
```

## Boxplot

```python
counter = 1
for c in train_dataset.columns:
    sns.boxplot(x=c, data=train_dataset, orient="h")
    plt.title(c)
    plt.show()
```

## Target pie

```python
plt.pie(x=train_dataset['target'].value_counts().tolist(), labels=train_dataset['target'].value_counts().index.tolist(), autopct="%.3f%%", colors=["red", "white"])
plt.title("Target Pie")
plt.show()
```
