# 2 - 
**Apparently it's easier to learn the orientation as a classification problem (eg. is the object facing forward/backward) then finetune the angle within the classified orientation than it is to just regress the angle.**

**Centernet splits the orientation angle in two overlapping 240¬∫ bins centered at -90¬∫ and 90¬∫. **

**For each bin, two scalars + softmax are used to predict if the angle falls into the bin, then two more scalars are used to encode the angle as sin(angle) and cos(angle).**

**thanks for your explanation @ Prof.Bacterio**

```python
import numpy as np
# get ùúÉùëô (Counter clockwise is positive)
def get_alpha(rot):
  # output: (Batch, 8) [bin1_cls[0], bin1_cls[1], bin1_sin, bin1_cos, 
  #                     bin2_cls[0], bin2_cls[1], bin2_sin, bin2_cos]
  # return rot[:, 0]
    idx = rot[:, 1] > rot[:, 5]
    
    # alpha1 is relative to -90¬∫
    alpha1 = np.arctan(rot[:, 2] / rot[:, 3]) + (-0.5 * np.pi)
    
    # alpha2 is relative to +90¬∫
    alpha2 = np.arctan(rot[:, 6] / rot[:, 7]) + ( 0.5 * np.pi)
    
    return alpha1 * idx + alpha2 * (1 - idx)
```
