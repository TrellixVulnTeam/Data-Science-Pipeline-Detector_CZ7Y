```python
rwood=[]
rcoal=[]
ruranium=[]

def get_wood_resources(game_state):
    global rwood
    global rcoal
    game_map=game_state.map
    (width, height) = (game_map.width, game_map.height)
    
    for i in range(width):
        for j in range(height):
            cell=game_map.get_cell(i, j)
            if cell.has_resource():
                resource=cell.resource
                amount=resource.amount
                resource_type=resource.type
                if resource_type == "wood":
                    rwood.append(amount)
                elif resource_type == 'coal':
                    rcoal.append(amount)
                else:
                    ruranium.append(amount)
```

```python
wood_amounts=[]
coal_amounts=[]
uranium_amounts=[]

for i in range(1000):
    if i%50 == 0:
        print(i)
    
    rwood=[]
    rcoal=[]
    ruranium=[]

    env=make("lux_ai_2021", 
             configuration={
                 "seed": random.randint(0, 10000000), 
                 "loglevel": 0,
                 "annotations": False,
                 "width": 12,
                 "height": 12
             },debug=True)

    _ = env.run([agent, 'simple_agent'])
    wood_amounts+=rwood
    coal_amounts+=rcoal
    uranium_amounts+=ruranium
```
