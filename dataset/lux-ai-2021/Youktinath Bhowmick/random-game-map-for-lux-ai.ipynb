```python
import numpy as np
import pandas as pd
import math
import sys
from lux.game import Game
from lux.game_map import Cell, RESOURCE_TYPES
from lux.constants import Constants
from lux.game_constants import GAME_CONSTANTS
from lux import annotate
import numpy as np
from collections import deque
import random, time
from IPython.display import clear_output
```

Next code shows a map for around 10 sec. The loop is for 1000 times, you can increase or decease as you wish.

```python
for _ in range(2):              # increase this value
    game_seed = math.floor(random.random() * 1e9)
    print(f"Showing map for seed value {game_seed}...")
    env = make("lux_ai_2021", configuration = {"seed": game_seed, "loglevel": 2, "annotations": True}, debug = True)
    steps = env.run(["simple_agent", "simple_agent"])
    env.render(mode = "ipython", width = 800, height = 600)
    time.sleep(13)
    clear_output()
    del env
    time.sleep(2)
    print("showing next map...")
    time.sleep(0.3)
    clear_output()
```
