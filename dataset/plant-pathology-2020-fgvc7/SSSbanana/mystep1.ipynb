```python
import os, shutil, random
import cv2
from PIL import Image, ImageEnhance
```

```python
count = 0 # 统计训练集文件数量
for f_name in os.listdir('../input/plant-pathology-2020-fgvc7/images'):
    if f_name.startswith('Train'):
        # 过滤训练集数据
        count=count + 1
print("The number of training set files is %d" % count)
```

```python
os.mkdir('./Train')
os.mkdir('./Validation')
os.mkdir('./Test')

for f_name in os.listdir('../input/plant-pathology-2020-fgvc7/images'):
    if f_name.startswith('Train'):
        num = random.randint(1, 10)    
        if num in range(1, 7):
            shutil.copy('../input/plant-pathology-2020-fgvc7/images/' + f_name, './Train')
            os.rename('./Train/' + f_name, './Train/' + f_name[6:])
        elif num in range(7, 9):
            shutil.copy('../input/plant-pathology-2020-fgvc7/images/' + f_name, './Validation')
            os.rename('./Validation/' + f_name, './Validation/' + f_name[6:])
        else:
            shutil.copy('../input/plant-pathology-2020-fgvc7/images/' + f_name, './Test')
            os.rename('./Test/' + f_name, './Test/' + f_name[6:])
```

```python
print("Training set : %d" % len(os.listdir("Train")))
print("Validation set : %d" % len(os.listdir("Validation")))
print("Test set : %d" % len(os.listdir("Test")))
```

```python
# 新建目录

os.mkdir('./A_Train')
os.mkdir('./A_Validation')
os.mkdir('./A_Test')

# 调整大小

def Adjust(src_dir,tar_dir):
    width = 256
    height = 256
    for f_name in os.listdir(src_dir):
        img = cv2.imread(src_dir + '/' + f_name)
        new_img = cv2.resize(img,(width,height))
        cv2.imwrite(tar_dir + '/' + f_name, new_img)
    
Adjust('./Train','./A_Train')
Adjust('./Validation','./A_Validation')
Adjust('./Test','./A_Test')
```

```python
# 滤镜

os.mkdir('./B_Train')
os.mkdir('./B_Validation')
os.mkdir('./B_Test')

def Nos(src_dir, tar_dir):
    for f_name in os.listdir(src_dir):
        img = Image.open(src_dir + '/' + f_name)
        
        img = ImageEnhance.Color(img).enhance(1.5)
        img = ImageEnhance.Brightness(img).enhance(1.5)
        img = ImageEnhance.Sharpness(img).enhance(1.5)
        img = ImageEnhance.Contrast(img).enhance(1.5)
        
        img.save(tar_dir + '/' + f_name)
    
Nos('./A_Train','./B_Train')
Nos('./A_Validation','./B_Validation')
Nos('./A_Test','./B_Test')
```

```python
os.mkdir('./Data')
os.mkdir('./Data/Train')
os.mkdir('./Data/Train/A')
os.mkdir('./Data/Train/B')
os.mkdir('./Data/Predict')
os.mkdir('./Data/Predict/Train')
os.mkdir('./Data/Predict/Validation')

for f_name in os.listdir("./A_Train"):
    shutil.copy('./A_Train/' + f_name, './Data/Train/A')
    shutil.copy('./A_Train/' + f_name, './Data/Predict/Train')
for f_name in os.listdir("./A_Validation"):
    shutil.copy('./A_Validation/' + f_name, './Data/Train/A')
    shutil.copy('./A_Validation/' + f_name, './Data/Predict/Validation')
for f_name in os.listdir("./B_Test"):
    shutil.copy('./B_Test/' + f_name, './Data/Train/B')
```
