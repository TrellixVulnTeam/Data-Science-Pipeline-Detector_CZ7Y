```python
import numpy as np 
import pandas as pd
import os
```

```python
img_path = '/kaggle/input/plant-pathology-2020-fgvc7/images/'
```

```python
df = pd.read_csv('/kaggle/input/plant-pathology-2020-fgvc7/train.csv')
display(df.head())

print(df.shape)
```

```python
df['image_id'] = df['image_id'].apply(lambda x: x + '.jpg')
```

```python
df['source'] = 2
```

```python
df['label']                                   = 4  
df.loc[df['rust'] == 1,              'label'] = 1
df.loc[df['scab'] == 1,              'label'] = 2
df.loc[df['multiple_diseases'] == 1, 'label'] = 3
df = df[['image_id', 'label', 'source']]
```

```python
print(df['label'].nunique())
df['label'].value_counts()
```

```python
from sklearn.model_selection import KFold, StratifiedKFold

df['fold'] = -1
skf = StratifiedKFold(n_splits = 5, random_state = 23, shuffle = True)
for fold, (trn_idx, val_idx) in enumerate(skf.split(df, df['label'])):
    df.loc[val_idx, 'fold'] = fold
```

```python
df.to_csv('df_2.csv', index = False)
df.head()
```
