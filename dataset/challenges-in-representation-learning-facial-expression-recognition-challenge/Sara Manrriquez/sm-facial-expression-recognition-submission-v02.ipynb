## Import Packages

```python
import numpy as np
import pandas as pd
import zipfile
import scikitplot

from tensorflow import keras
from sklearn.metrics import classification_report
```

## Load Test DataFrame

```python
test = pd.read_csv('../input/challenges-in-representation-learning-facial-expression-recognition-challenge/test.csv')
print(test.shape)
```

```python
test.head()
```

## Preprocess Data

```python
test['pixels'] = [np.fromstring(x, dtype=int, sep=' ').reshape(-1,48,48,1) for x in test['pixels']]
```

```python
pixels = np.concatenate(test['pixels'])

print(pixels.shape)
```

```python
pixels = pixels / 255
```

## Load Model

```python
cnn = keras.models.load_model('../input/sm-facial-expression-recognition-v03/fer_model_v03.h5')
cnn.summary()
```

## Test Predictions

```python
test_probs = cnn.predict(pixels)
print(test_probs.shape)
```

```python
print(test_probs[:10,].round(2))
```

```python
test_pred = np.argmax(test_probs, axis=1)
print(test_pred[:10])
```

## Evaluation

```python
test_labels = pd.read_csv('../input/challenges-in-representation-learning-facial-expression-recognition-challenge/icml_face_data.csv')
test_labels.columns = ['emotion', 'Usage', 'pixels']
print(test_labels.shape)
```

```python
test_labels = test_labels.loc[test_labels['Usage'].isin(['PublicTest', 'PrivateTest'])]
```

```python
test_labels.drop(columns='Usage', inplace=True)
test_labels.head()
```

```python
test_labels.shape
```

```python
test_labels['predictions'] = test_pred
test_labels.head()
```

```python
my_classification_report = classification_report(test_labels['emotion'], test_labels['predictions'])
print(my_classification_report)
```

```python
print('Total Wrong Predictions:', np.sum(test_labels['emotion'] != test_labels['predictions']))
```

```python
scikitplot.metrics.plot_confusion_matrix(test_labels['emotion'], test_labels['predictions'], figsize=(7,7))    
```
