# HCD Submission

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import zipfile
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

# Load Dataframe

```python
test = pd.read_csv('../input/histopathologic-cancer-detection/sample_submission.csv')

print('Test Set Size:', test.shape)
```

```python
test['filename'] = test.id + '.tif'
```

```python
test.head()
```

```python
test_path = "../input/histopathologic-cancer-detection/test"
print('Test Images:', len(os.listdir(test_path)))
```

# Data Generator

```python
BATCH_SIZE = 64

test_datagen = ImageDataGenerator(rescale=1/255)

test_loader = test_datagen.flow_from_dataframe(
    dataframe = test,
    directory = test_path,
    x_col = 'filename',
    batch_size = BATCH_SIZE,
    shuffle = False,
    class_mode = None,
    target_size = (96,96)
)
```

# Load Model

```python
cnn = keras.models.load_model('../input/hcd601/HCDv01.h5')
cnn.summary()
```

```python
test_probs = cnn.predict(test_loader)
print(test_probs.shape)
```

```python
print(len(test_loader))
```

```python
print(test_probs[:10,].round(2))
```

```python
test_pred = np.argmax(test_probs, axis=1)
print(test_pred[:10])
```

# Prepare Submission

```python
submission = pd.read_csv('../input/histopathologic-cancer-detection/sample_submission.csv')
submission.head()
```

```python
submission.label = test_probs[:,1]
submission.head()
```

```python
submission.to_csv('submission.csv', header=True, index=False)
```
