```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.image as mpimg

test_path = '/kaggle/input/histopathologic-cancer-detection/test/'
```

```python
test = pd.read_csv(f'../input/histopathologic-cancer-detection/sample_submission.csv', dtype=str)

print('Test Set Size:', test.shape)

test.head()
```

```python
test['path'] = test.id + '.tif'
```

```python
test_datagen = ImageDataGenerator(rescale=1/255)
```

```python
BATCH_SIZE = 64

test_loader = test_datagen.flow_from_dataframe(
    dataframe = test,
    directory = test_path,
    x_col = 'path',
    batch_size = BATCH_SIZE,
    shuffle = False,
    class_mode = None,
    target_size = (96,96)
)
```

```python
cnn = keras.models.load_model('../input/modelsaug3/JW2_cancer_model_v01.h5')
cnn.summary()
```

```python
test_probs = cnn.predict(test_loader)
```

```python
print(test_probs[:10,].round(2))
```

```python
submission = pd.read_csv(f'../input/histopathologic-cancer-detection/sample_submission.csv', dtype=str)
submission.label = test_probs[:,1]
submission.head()
```
