# Import Statements

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import matplotlib.image as mpimg

from sklearn.model_selection import train_test_split

import pickle

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras.preprocessing.image import ImageDataGenerator

import os
```

```python
test = pd.read_csv('../input/histopathologic-cancer-detection/sample_submission.csv')
print('Test Set Size:', test.shape)

# test = test.iloc[:100,]
# print('Test Set Size:', test.shape)
```

```python
test.head()
```

```python
test['id'] = test['id'].apply(lambda x: f'{x}.tif')
test.head()
```

# Data Generator

```python
BATCH_SIZE = 64
test_path = "../input/histopathologic-cancer-detection/test"
print('Test Images:', len(os.listdir(test_path)))

test_datagen = ImageDataGenerator(rescale=1/255)

test_loader = test_datagen.flow_from_dataframe(
    dataframe = test,
    directory = test_path,
    x_col = 'id',
    batch_size = BATCH_SIZE,
    shuffle = False,
    class_mode = None,
    target_size = (96,96)
)
```

```python
cnn = keras.models.load_model('../input/cancer-models/cancer_model_v01_version10.h5')
cnn.summary()
```

# Test Predictions

```python
test_probs = cnn.predict(test_loader)
print(test_probs.shape)
```

```python
print(test_probs[:10,].round(2))
```

# Update submission

```python
submission = pd.read_csv('../input/histopathologic-cancer-detection/sample_submission.csv')
submission.head()
```

```python
submission.label = test_probs[:,1]
submission.head()
```

```python
submission.to_csv('submission.csv', header=True, index=False)
```
