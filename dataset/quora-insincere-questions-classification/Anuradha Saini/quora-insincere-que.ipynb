```python
import string
string.punctuation
```

```python
#defining the function to remove punctuation
def remove_punctuation(text):
    punctuationfree="".join([i for i in text if i not in string.punctuation])
    return punctuationfree

target_1['clean_msg']= target_1['question_text'].apply(lambda x:remove_punctuation(x))
target_1.head()
```

```python
# target_1.columns.values
target_1['msg_lower']= target_1['clean_msg'].apply(lambda x: x.lower())
target_1.iloc[:5,[3,4]]
```

```python
#defining function for tokenization
import re
def tokenization(text):
    tokens = re.split('\W+',text)
    return tokens
#applying function to the column
target_1['msg_tokenied']= target_1['msg_lower'].apply(lambda x: tokenization(x))
target_1.iloc[:3,-1:]
# text = 'if blacks support school choicey vote republican'
# tokens = re.split('\W+',text)
# tokens
# import nltk
# word_data = 'if blacks support school choicey vote republican'
# nltk_tokens = nltk.word_tokenize(word_data)
# print (nltk_tokens)
```

```python
#importing nlp library
import nltk
#Stop words present in the library
stopwords = nltk.corpus.stopwords.words('english')
stopwords[0:10]
```

```python
#defining the function to remove stopwords from tokenized text
def remove_stopwords(text):
    output= [i for i in text if i not in stopwords]
    return output

target_1['stpw_remov'] = target_1['msg_tokenied'].apply(lambda x : remove_stopwords(x))
target_1.iloc[:3,-2:]
# target_1.head(2)
```

```python
from nltk.stem.porter import PorterStemmer
porter_stemmer = PorterStemmer()
```

```python
def stemming(text):
    stem_text = [porter_stemmer.stem(word) for word in text]
    return stem_text

target_1['msg_stemmed']=target_1['stpw_remov'].apply(lambda x: stemming(x))
target_1.iloc[:3,-2:]
```

```python
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()

def lemmatizer(text):
    lemm_text = [wordnet_lemmatizer.lemmatize(word) for word in text]
    return lemm_text

target_1['msg_lemmatized']=target_1['stpw_remov'].apply(lambda x:lemmatizer(x))
target_1.iloc[:3,-3:]
```
