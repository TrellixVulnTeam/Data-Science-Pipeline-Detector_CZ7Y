```python
train_df = pd.read_csv('../input/quora-insincere-questions-classification/train.csv')
test_df = pd.read_csv('../input/quora-insincere-questions-classification/test.csv')

train_X = train_df['question_text']
test_X = test_df['question_text']
```

```python
from keras.preprocessing.text import Tokenizer


tokenizer = Tokenizer(num_words = 150000)

#tokenizer từ trong tập train và tập huấn luyện
tokenizer.fit_on_texts(list(train_X)+list(test_X))

# tạo một dict để mapping từ và số thứ tự
word_index = tokenizer.word_index # type: dict
list_word = list(word_index.keys())


```

# Tìm các từ contraction

```python
list_contraction = []
for word in list_word:
    if "'" in word:
        list_contraction.append(word)
print(list_contraction)
```

Ta có thể thấy các từ contration là viết tắt quen thuộc như I'm = I am, I'll = I will,..... ta sẽ xử lý những từ thông dụng như này
