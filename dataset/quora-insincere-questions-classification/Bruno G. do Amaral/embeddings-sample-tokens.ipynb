```python
def sample_tokens(fname, n=200, encoding='utf-8'):
    with codecs.open(fname, 'r', encoding=encoding) as f:
        lines = []

        for i in range(n):
            lines.append(f.readline().split(' ', 2)[0])
    return lines
```
