```python
import pandas as pd
df_train = pd.read_csv("../input/train.csv")
df_test = pd.read_csv("../input/test.csv")
df_train.head()
```

```python
df = df_train[['qid', 'question_text']].append(df_test)
```

```python
df[df.duplicated(['qid'], keep=False)]
```

```python
for i in range(1, 21):
    df_train['first_' + str(i)] = df_train['qid'].apply(lambda x: x[:i])
    
df_train.head()
```
