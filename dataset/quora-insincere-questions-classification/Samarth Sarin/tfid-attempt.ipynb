```python
df = pd.read_csv("../input/train.csv")
```

```python
df.head()
```

```python
df.drop('qid',1,inplace = True)
```

```python
lem = WordNetLemmatizer()
```

```python
def cleaning(text):
    words = []
    doc = nlp(text)
    for token in doc:
        if not token.is_punct:
            words.append(lem.lemmatize(token.text))
    return ' '.join(words)
    
```

```python
df['question_text'] = df['question_text'].apply(cleaning)
```
