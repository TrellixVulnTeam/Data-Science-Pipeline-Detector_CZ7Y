```python
import os
import pandas as pd
import matplotlib.pyplot as plt

df_train = pd.read_csv(os.path.join(PATH_DATASET, "train.csv"))
display(df_train.head())
print(f"Dataset size: {len(df_train)}")
print(f"Unique ids: {len(df_train['individual_id'].unique())}")
```

Lets see how top individulas we have in the database...

```python
top_ = df_train["individual_id"].value_counts()
ax = top_[:50].plot.barh(figsize=(3, 8), grid=True)  # ascending=True
```

```python
display(top_[:4])
top_ids = list(top_[:4].index) + ['new_individual']
print(top_ids)
```

```python
import glob

imgs = glob.glob("/kaggle/input/happy-whale-and-dolphin/test_images/*.jpg")
imgs = list(map(os.path.basename, imgs))
print(imgs[:5])
```
