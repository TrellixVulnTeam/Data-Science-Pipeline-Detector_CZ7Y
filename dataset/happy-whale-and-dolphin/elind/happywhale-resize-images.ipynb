```python
import os
from PIL import Image
from tqdm import tqdm
import shutil
```

```python
target_size = 224
target_size
```

```python
directory = "../input/happy-whale-and-dolphin/train_images"
filenames = os.listdir(directory)
filenames[:5], len(filenames)
```

```python
#filenames = filenames[:99]
```

```python
folder = "happywhale_resized_images"
os.makedirs(folder, exist_ok=True)

for i, filename in tqdm(enumerate(filenames), total=len(filenames)):
    filepath = os.path.join(directory, filename)
    image = Image.open(filepath)
    original_image_size = image.size
    image_size = tuple(
        [
            round(dimension * target_size / max(original_image_size))
            for dimension in original_image_size
        ]
    )
    image = image.resize(image_size)
    new_image = Image.new("RGB", (target_size, target_size))
    new_image.paste(
        image, ((target_size - image_size[0]) // 2, (target_size - image_size[1]) // 2)
    )
    new_filepath = os.path.join(folder, filename)
    new_image.save(new_filepath)

    if i < 9:
        display(new_image.size)
        display(new_image)
```
