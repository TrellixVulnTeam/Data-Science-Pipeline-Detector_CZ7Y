```python
import numpy as np
import pandas as pd
import os
import cv2
from tqdm import tqdm
tqdm.pandas()
```

```python
path = "../input/happywhale-enhanced-dataset-light"

def is_greyscale(folder):
    def f(x):
        image_path = os.path.join(path, folder, x)
        image = cv2.imread(image_path)[:, :, ::-1]
        return (image[:, :, 0] == image[:, :, 1]).all() and (image[:, :, 1] == image[:, :, 2]).all()
    return f
```

```python
data = pd.read_csv(os.path.join(path, "train.csv"))
greyscale_map = data["image"].progress_map(is_greyscale("train_images"))
np.save("train_grey_scale_mask.npy", greyscale_map.values)
```

```python
data = pd.read_csv(os.path.join(path, "sample_submission.csv"))
greyscale_map = data["inference_image"].progress_map(is_greyscale("test_images"))
np.save("test_grey_scale_mask.npy", greyscale_map.values)
```
