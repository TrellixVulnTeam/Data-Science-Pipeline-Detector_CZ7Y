```python
import numpy as np
import pandas as pd
import os
import cv2
import matplotlib.pyplot as plt
```

```python
def plot_bbox(image, x1, y1, x2, y2):
    path = os.path.join("../input/happy-whale-and-dolphin/train_images", image)
    image = cv2.imread(path)
    xmin = int(min(x1, x2) * image.shape[1])
    ymin = int(min(y1, y2) * image.shape[0])
    xmax = int(max(x1, x2) * image.shape[1])
    ymax = int(max(y1, y2) * image.shape[0])
    image = cv2.rectangle(
        image,
        (xmin, ymin),
        (xmax, ymax),
        (0,255,255),
        3
    )
    plt.imshow(image[:, :, ::-1])
    
def plot_cropped(image, x1, y1, x2, y2):
    path = os.path.join("../input/happy-whale-and-dolphin/train_images", image)
    image = cv2.imread(path)
    xmin = int(min(x1, x2) * image.shape[1])
    ymin = int(min(y1, y2) * image.shape[0])
    xmax = int(max(x1, x2) * image.shape[1])
    ymax = int(max(y1, y2) * image.shape[0])
    plt.imshow(image[ymin:ymax, xmin:xmax, ::-1])
    
def plot_bbox_dic(x):
    plot_bbox(
        x["image"],
        x["x1"],
        x["y1"],
        x["x2"],
        x["y2"]
    )
    
def plot_cropped_dic(x):
    plot_cropped(
        x["image"],
        x["x1"],
        x["y1"],
        x["x2"],
        x["y2"]
    )
```

```python
data = pd.read_csv("../input/wandd-crowed-sourced-bounging-boxes/dataset.csv")
```

```python
N_SAMPLES = 10
```

```python
col=2
row=N_SAMPLES
plt.figure(figsize=(col*6, row*2))
samples = data.sample(N_SAMPLES)
for i in range(row*col):
    plt.subplot(row, col, i+1)
    
    x = samples.iloc[i//2]
    if i%2 == 0:
        plt.title(x["image"])
        plot_bbox_dic(x)
    else:
        plt.title(x["image"])
        plot_cropped_dic(x)
        
    plt.axis('off')
plt.tight_layout()
plt.show()
```
