```python
# Just gets the shape pretty quick, nothing more!

import joblib
import os
from skimage import io
import pandas as pd
from joblib import Parallel, delayed
```

```python
%%time
path = "/kaggle/input/happy-whale-and-dolphin/"
df = pd.read_csv("/kaggle/input/happy-whale-and-dolphin/train.csv")

def process(i):
    im = io.imread(path+'train_images/'+ df.iloc[i].image)
    shape = list(im.shape)
    return shape if len(shape) == 3 else shape+[1]
    
df['shape'] = Parallel(n_jobs=4)(delayed(process)(i) for i in range(len(df)))
df[['d0','d1','d2']] = pd.DataFrame(df['shape'].to_list())
df.drop(columns='shape', inplace=True)
df.to_csv("/kaggle/working/train_extended.csv", index=False)
```

```python
%%time
path = "/kaggle/input/happy-whale-and-dolphin/"
df = pd.read_csv("/kaggle/input/happy-whale-and-dolphin/sample_submission.csv")

def process(i):
    im = io.imread(path+'test_images/'+ df.iloc[i].image)
    shape = list(im.shape)
    return shape if len(shape) == 3 else shape+[1]
    
df['shape'] = Parallel(n_jobs=4)(delayed(process)(i) for i in range(len(df)))
df[['d0','d1','d2']] = pd.DataFrame(df['shape'].to_list())
df.drop(columns='shape', inplace=True)
df.to_csv("/kaggle/working/sample_submission_extended.csv", index=False)
```
