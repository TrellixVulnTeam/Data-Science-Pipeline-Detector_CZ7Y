```python
import pandas as pd
df = pd.read_csv("../input/happy-whale-and-dolphin/train.csv")
```

```python
# Get the train image file names
# Use the * as a wild card this will tell glob get us all images in this directory
image_path = '/kaggle/input/jpeg-happywhale-128x128/train_images-128-128/train_images-128-128/*'
train_filenames = glob.glob(image_path)
```

```python
# Get a single image path
image_path = train_filenames[67]
img = cv2.imread(image_path)

# Check the shape - should be (128, 128, 3)
img.shape
```

```python
transform = A.Compose([
    A.RandomCrop(width=128, height=128),
    A.HorizontalFlip(p=0.5),
    A.RandomBrightnessContrast(p=0.2),
])
```

```python
pillow_image = Image.open(train_filenames[678])
image = np.array(pillow_image)
```

```python
transformed = transform(image=image)
```

```python
plt.axis('off')
plt.imshow(transformed['image'])
plt.show()
```

### Color Augmentations


```python
light = A.Compose([
    A.RandomBrightnessContrast(),    
    A.RandomGamma(p=1),    
], p=1)
```

```python
plt.axis('off')
plt.imshow(light(image=image)['image'])
plt.show()
```

```python
hue = A.Compose([
    A.HueSaturationValue(hue_shift_limit=20, sat_shift_limit=50, val_shift_limit=50, p=1),
], p=1)
```

```python
plt.axis('off')
plt.imshow(hue(image=image)['image'])
plt.show()
```

```python
strong = A.Compose([
    A.ChannelShuffle(p=0.2),
], p=1)
```

```python
plt.axis('off')
plt.imshow(strong(image=image)['image'])
plt.show()
```

```python
snow = A.Compose(
    [A.RandomSnow(brightness_coeff=2.5, snow_point_lower=0.3, snow_point_upper=0.5, p=1)],
)
```

```python
plt.axis('off')
plt.imshow(snow(image=image)['image'])
plt.show()
```
