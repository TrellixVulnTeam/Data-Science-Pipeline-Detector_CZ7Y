## 必要なライブラリのインポート

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import os

import cv2
```

```python
train_data = pd.read_csv("../input/happy-whale-and-dolphin/train.csv")
train_data
```

## 個体別に表示

- まず個体別にみてみます

```python
idval_data = train_data.groupby('individual_id')['image'].count().sort_values(ascending=True)
idval_data
```

- 上の結果だけでは分かりにくいので、1枚の画像しかない個体の数、2枚の画像しかない個体の数、3枚から5枚の画像がある個体の数...といった感じで、画像の枚数ごとに分けてみます

```python
idval_ctgr_list = [0,1,2,5,20,100,1000]
idval_ctgr_name = ['1','2','3-5','6-20','21-100','100-']
idval_ctgr = pd.cut(idval_data, bins=idval_ctgr_list, labels=idval_ctgr_name)
idval_ctgr
```

- ほとんどの個体が、学習データとして1枚または2枚の画像しかもたないことが分かります

```python
plot_data = idval_ctgr.value_counts().sort_index()
plt.bar(plot_data.index, plot_data)
```

## 種類別の画像枚数を表示

- 次にtrain_imagesの中に、各種類ごとに何枚ずつのデータがあるのかを調べてみます

```python
species_count = train_data.groupby("species")["image"].count()
species_count = species_count.sort_values(ascending=False)
print(species_count)
species_count.plot(kind="bar")
```

```python
species_name_list = species_count.index
species_name_list
```

## sample画像の表示

- train_imagesには全部で30種類の種の画像があります. 各種類の画像を見てみましょう.

```python
def display_sample_image(species_num=0):
    """
    species_numで0から29の数値を指定すると、対応する種の画像がランダムに4枚表示されます。
    """
    species_num %= 29
    base_path = '../input/happy-whale-and-dolphin/train_images'
    fig = plt.figure(figsize=(12,8))
    plt.title(species_name_list[species_num])
    plt.yticks([])
    plt.xticks([])
    
    train_species_data = train_data[train_data.species==species_name_list[species_num]].reset_index(drop=True)
    for i in range(4):
        index = np.random.randint(len(train_species_data))
        image_path = os.path.join(base_path, train_species_data.loc[index]["image"])
        sample_pic = cv2.imread(image_path)
        
        ax = fig.add_subplot(2,2,i+1)
        ax.imshow(sample_pic)
#     plt.tight_layout()
```
