```python
import matplotlib.pyplot as plt
import os
import random
```

# Get the path of each image

```python
def getPaths(basePath):
    paths = []
    for dirname, _, filenames in os.walk(basePath):
        for filename in filenames:
            path = os.path.join(dirname, filename)
            paths.append(path)
    return paths
```

```python
trainImagesPaths = getPaths("/kaggle/input/happy-whale-and-dolphin/train_images")
```

# View images

Images have a different shape and resolution!

```python
def showImages(paths, nrows, ncols, height, width, seed = 0, dpi = 200):
    fig = plt.figure(figsize = (width, height), dpi = dpi)
    for i in range(0, nrows * ncols):
        image = plt.imread(paths[i])
        axis = fig.add_subplot(nrows, ncols, i + 1)
        axis.set_axis_off()
        plt.imshow(image)
    plt.tight_layout()
    plt.show()
```

```python
nrows = 5
ncols = 5
height = 10
width = 10
images = random.choices(trainImagesPaths, k = nrows * ncols)
showImages(images, nrows, ncols, height, width)
```
