# Taxonomy

The purpose of this notebook is to join the table `train.csv` with a taxonomy table.

```python
import pandas as pd
taxonomy = pd.read_csv("/kaggle/input/happywhalespeciesclassification/species.csv")
taxonomy.shape
```

```python
#taxonomy.sort_values(by = ['infraorder','family','genus','specy']).drop(columns = ['wikipedia','image','size']).set_index('specy_id')
taxonomy = taxonomy.sort_values(by = ['specy']).drop(columns = ['wikipedia','image','size']).set_index('specy_id')
```

```python
taxonomy
```

```python
df = pd.read_csv("/kaggle/input/happy-whale-and-dolphin/train.csv")
df.loc[df.species == 'bottlenose_dolpin', 'species'] = 'bottlenose_dolphin'
df.loc[df.species == 'kiler_whale', 'species'] = 'killer_whale'
df.loc[df.species == 'long_finned_pilot_whale', 'species'] = 'pilot_whale'
df.loc[df.species == 'globis', 'species'] = 'pilot_whale'
df = df.rename(columns = { 'species': 'specy_id' })
final = pd.merge(df, taxonomy, on = ['specy_id'], how = 'left')
final
```

```python
final.to_csv("/kaggle/working/final.csv", index = False)
```
