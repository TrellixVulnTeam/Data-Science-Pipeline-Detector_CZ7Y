# Let's find the number of images per specie

# The number of image per species seems to be inbalanced. üßê

# Explore some images

```python
df_show = df.sample(9)
img_ids = df_show['image'].values
species_ids = df_show['species'].values

path = '../input/happy-whale-and-dolphin/train_images/'

fig = plt.figure(figsize=(15, 15))
columns = 3
rows = 3
for i in range(1, columns*rows +1):
    img = cv2.imread(path+ img_ids[i-1])
    fig.add_subplot(rows, columns, i)
    plt.imshow(img)
    plt.title(species_ids[i-1])
plt.show()
```

# Now by unique id which is the label we will predict.

```python
random_id = df.individual_id.sample(1).values
unique_id = df[df['individual_id'] == random_id[0]]

img_ids = unique_id.sample(3)['image'].values

path = '../input/happy-whale-and-dolphin/train_images/'

fig = plt.figure(figsize=(15, 15))
columns = 3
rows = 1
for i in range(1, columns*rows +1):
    img = cv2.imread(path+ img_ids[i-1])
    fig.add_subplot(rows, columns, i)
    plt.imshow(img)
    plt.title(unique_id.species.unique()[0] + '\n' + random_id[0])
plt.show()
```
