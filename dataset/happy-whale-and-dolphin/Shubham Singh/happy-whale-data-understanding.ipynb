```python
train['species'] = train['species'].str.replace('bottlenose_dolpin','bottlenose_dolphin')
train['species'] = train['species'].str.replace('kiler_whale','killer_whale')
```

```python
# append _whale to beluga and globis
train["species"].replace({"beluga": "beluga_whale", "globis": "globis_whale"}, inplace=True)
```

```python
(train.species.value_counts()).plot.bar(figsize=(12,8))
plt.xticks(fontsize=12,fontweight='bold')
plt.yticks(fontsize=12,fontweight='bold')
plt.show()
```

```python
train['label'] = train.species.map(lambda x: 'dolphin' if 'dolphin' in x else 'whale')

fig, ax  = plt.subplots(figsize=(16, 8))
fig.suptitle('Whales and Dolphins ', size = 20, font="Serif")
explode = (0.05, 0.05)
labels = list(train.label.value_counts().index)
sizes = train.label.value_counts().values
ax.pie(sizes, explode=explode,startangle=60, labels=labels,autopct='%1.0f%%', pctdistance=0.7, colors=["#0077b6","#90e0ef"])
ax.add_artist(plt.Circle((0,0),0.4,fc='white'))
plt.show()
```

```python
def show_images(nrows,ncols,data):
    figure, axs= plt.subplots(nrows,ncols,figsize=(16,8))
    for i,ax in enumerate(axs.ravel()):
        img=cv2.imread(data[i])
        img=cv2.resize(img,(300,300))
        img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
        ax.imshow(img,interpolation=None)
        plt.tight_layout()
    plt.show()
```

```python
#Display train images
show_images(5,3,img_path_train)
```

```python
# Display test images
show_images(5,3,img_path_test)
```
