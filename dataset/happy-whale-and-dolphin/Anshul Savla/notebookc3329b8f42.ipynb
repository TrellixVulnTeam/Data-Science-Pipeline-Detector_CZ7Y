```python
import numpy as np
import pandas as pd
import os
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D, BatchNormalization
from tensorflow.keras.optimizers import RMSprop
import keras
import keras.utils
from keras import utils as np_utils
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
def read_img(folder):
    imgg=[]
    for filename in os.listdir(folder):
        img = plt.imread(os.path.join(folder , filename))
        #img = cv2.resize(img , (200 , 200))
        imgg.append(img)
    return imgg
```

```python
train = '../input/jpeg-happywhale-128x128/train_images-128-128/train_images-128-128'
test = '../input/jpeg-happywhale-128x128/test_images-128-128/test_images-128-128'
train_df = '../input/happy-whale-and-dolphin/train.csv'
```

```python
from tensorflow.keras.utils import to_categorical
y_train = to_categorical(y_train)
```

```python
y_train.shape
```

```python
X_train = np.array(X_train)
X_test = np.array(X_test)
```

```python
cp = keras.callbacks.ModelCheckpoint(filepath = './weights' , monitor='accuracy' , save_best_only=True)
es = keras.callbacks.EarlyStopping(monitor='loss',patience=3)
```

```python
inputs = keras.Input(shape=(X_train.shape[1],X_train.shape[2],X_train.shape[3]))
x = Conv2D(64, (3, 3), activation='relu', padding='same')(inputs)
x = Conv2D(32, (3, 3), activation='relu', padding='same',)(x)
x = MaxPooling2D(pool_size=(2, 2))(x)
x = Dropout(0.2)(x)
x = Conv2D(64, (3, 3), activation='relu', padding='same')(x)
x = Conv2D(32, (3, 3), activation='relu', padding='same',)(x)
x = MaxPooling2D(pool_size=(2, 2))(x)
x = Conv2D(64, (3, 3), activation='relu', padding='same')(x)
x = Conv2D(32, (3, 3), activation='relu', padding='same',)(x)
x = MaxPooling2D(pool_size=(2, 2))(x)
x = Flatten()(x)
x = Dense(128,activation='relu')(x)
x = Dense(64,activation='relu')(x)
x = Dense(32,activation='relu')(x)
output = Dense(30, activation='softmax')(x)
model = keras.Model(inputs, output)

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```
