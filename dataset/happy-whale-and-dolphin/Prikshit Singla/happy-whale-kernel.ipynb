# Exploratory Data Analysis

### Let's explore the species first!

```python

print(len(train_df['species'].value_counts()))
train_df['species'].value_counts()
```

#### So there are 30 species in total in the training images
<ul>
<li>Beluga is a whale species here
<li>Globis is also a whale specie
<li>There are duplicates so we have to remove the duplicate species

```python
print(colored('Before fixing duplicate labels : ','green'))
print("Number of unique species : ", train_df['species'].nunique())
print("\nSpecies names:" ,train_df['species'].unique())


#fix duplicates
train_df['species']  = train_df['species'].str.replace('bottlenose_dolpin','bottlenose_dolphin')
train_df['species']  = train_df['species'].str.replace('kiler_whale','killer_whale')
print(colored('\nAfter fixing duplicate labels : ','green'))
print("Number of unique species : ", train_df['species'].nunique())
print("\nSpecies names:" ,train_df['species'].unique())

train_df['species'].replace({'beluga' : 'beluga_whale','globis' : 'globis_whale'},inplace =True)

```

```python
plt.figure(figsize = (7,7))
sns.countplot(data = train_df , y = 'species',palette = 'plasma');
```

### The data contains the pictures of bottlenose dolphins the highest

```python
train_df['D_or_W'] = ([i[-1] for i in train_df['species'].str.split('_')])
train_df['D_or_W'].head()
```

```python
explode = (0.05, 0.05)
plt.pie(x  = train_df['D_or_W'].value_counts(),labels = ['Whale','Dolphin'],colors=["#0077b6","#90e0ef"],
        textprops={'fontsize': 13},autopct = '%1.1f%%',explode = explode)
plt.show()
```

### So there are more Whales in the dataset as compared to Dolphins

```python
### Show one image for each species
plt.figure(figsize = (15,12))
for idx,i in enumerate(train_df.species.unique()):
    plt.subplot(4,7,idx+1)
    df = train_df[train_df['species'] ==i].reset_index(drop = True)
    image_path = df.loc[0,'Path']
    img = Image.open(image_path)
    img = img.resize((224,224))
    plt.imshow(img)
    plt.axis('off')
    plt.title(i)
plt.tight_layout()
plt.show()
```

```python
def plot_species(df,species_name):
    plt.figure(figsize = (10,10))
    species_df = df[df['species'] ==species_name].reset_index(drop = True)
    plt.suptitle(species_name)
    for idx,i in enumerate(np.random.choice(species_df['Path'],9)):
        plt.subplot(3,3,idx+1)
        image_path = i
        img = Image.open(image_path)
        img = img.resize((224,224))
        plt.imshow(img)
        plt.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
plot_species(train_df , 'pygmy_killer_whale')
```

```python
plot_species(train_df , 'gray_whale')
```

### Let's see the individual id's now!

```python
print("Number of unique individual whales and dolphins : ",len(train_df['individual_id'].value_counts()))
train_df['individual_id'].value_counts()
```

```python
def plot_individual(df,individual_id):
    plt.figure(figsize = (10,10))
    species_df = df[df['individual_id'] ==individual_id].reset_index(drop = True)
    plt.suptitle("ID : {} specie : {}".format(individual_id,species_df.species[0]))
    for idx,i in enumerate(np.random.choice(species_df['Path'],9)):
        plt.subplot(3,3,idx+1)
        image_path = i
        img = Image.open(image_path)
        img = img.resize((224,224))
        plt.imshow(img)
        plt.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
plot_individual(train_df , '37c7aba965a5')
```

```python
plot_individual(train_df , 'a6e325d8e924')
```
