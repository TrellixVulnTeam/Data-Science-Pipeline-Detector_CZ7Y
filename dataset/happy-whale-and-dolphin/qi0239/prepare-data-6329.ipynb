```python
import pandas as pd
import numpy as np
```

```python
kflod_10 = pd.read_csv('../input/happywhale-splits/skf_species_10folds.csv')
train_df = pd.read_csv('../input/happy-whale-and-dolphin/train.csv')
```

```python
train_df["individual_id_num"] = kflod_10["individual_id"]
train_df["species_num"] = kflod_10["species"]
```

```python
map_dict_spe = train_df[["species", "species_num"]].set_index("species").to_dict(orient='dict')['species_num']
map_dict_ind = train_df[["individual_id", "individual_id_num"]].set_index("individual_id").to_dict(orient='dict')['individual_id_num']
```

```python
train_df['nums']=train_df['individual_id'].map(dict(train_df['individual_id'].value_counts()))
```

```python
train_df = train_df[train_df['nums']>1]
```

```python
train_df = train_df.reset_index(drop = True)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold
import joblib
```

```python
encoder = LabelEncoder() # 6329
train_df['individual_id_num'] = encoder.fit_transform(train_df['individual_id'])
```

```python
with open("./le_6329.pkl", "wb") as fp:
    joblib.dump(encoder, fp)
```

```python
skf = StratifiedKFold(n_splits=10)
for fold, ( _, val_) in enumerate(skf.split(X=train_df, y=train_df.individual_id_num)):
      train_df.loc[val_ , "kfold"] = fold
```

```python
train_df.to_csv('./train_fold10_onlymore2_6329.csv',index=False)
```
