# What a problem?

I see in kernels some whale vs dolphin distribution ([one](https://www.kaggle.com/lokesharya99/happy-whale-and-dolphin-an-eda), [two](https://www.kaggle.com/gpreda/happy-whales-and-dolphins)). All of them are wrong :( The truth is: if specie has `_whale` in name, it doesn't mean it is the whale! 

# Who is who?

So, good news: all `_dolphin` are dolphins, `beluga` is whale, but some whales are dolphins too: `false_killer_whale`, `killer_whale`, `pygmy_killer_whale`, `long_finned_pilot_whale`, `short_finned_pilot_whale`, and `melon_headed_whale`. 

So, at total **10 whales and 16 dolphins species**.

See details about species in my previous [kernel](https://www.kaggle.com/kwentar/what-about-species)

# Correct distributions:

```python
import pandas as pd
train_df = pd.read_csv('../input/happy-whale-and-dolphin/train.csv')
train_df.species.replace({"globis": "short_finned_pilot_whale",
                          "pilot_whale": "short_finned_pilot_whale",
                          "kiler_whale": "killer_whale",
                          "bottlenose_dolpin": "bottlenose_dolphin"}, inplace=True)
```

```python
whale_species = ['humpback_whale', 'beluga', 'blue_whale', 'fin_whale', 'southern_right_whale',
                'gray_whale', 'minke_whale', 'cuviers_beaked_whale', 'sei_whale', 'brydes_whale']

train_df['whale_dolphin'] = ['whale' if x in whale_species else 'dolphin' for x in train_df.species]
train_df.head()
```

```python
import plotly.express as px
print(train_df.whale_dolphin.value_counts())
fig = px.pie(train_df.whale_dolphin, 
             values=train_df.whale_dolphin.value_counts().values, 
             names=train_df.whale_dolphin.value_counts().index)
fig.show()
```
