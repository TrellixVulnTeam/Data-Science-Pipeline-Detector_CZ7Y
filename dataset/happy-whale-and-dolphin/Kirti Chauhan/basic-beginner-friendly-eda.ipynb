Let's start the visualisation party!!

```python
import cv2
sample = train_df[train_df.species == 'humpback_whale'].sample(9)
plt.figure(figsize=(12,12))
for ind, (image_id, label) in enumerate(zip(sample.image, sample.species)):
    plt.subplot(3, 3, ind + 1)
    image = cv2.imread(os.path.join("../input/happy-whale-and-dolphin/train_images", image_id))
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image)
    plt.axis("off")
    
plt.show()
```

These are some happy humback whales.
