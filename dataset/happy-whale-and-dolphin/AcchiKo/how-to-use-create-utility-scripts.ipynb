# <B>[2] Creation of utility scripts</B>

## [2-1] Creation of new notebook

Creates new notebook for defining utility scripts.

## [2-2] Setting as utility script

Sets the notebook as utility script by clicking the following items in the menu of notebook,

### "File"  -->  "Set as Utility Script".

## [2-3] Setting editor type

Sets editor type by clicking the following items in the menu of notebook,

### "File"  -->  "Editor Type"  -->  "Script".

## [2-4] Defining fuctions/classes

Defines required functions/classes in the notebook.

## [2-5] Saving notebook

Saves the notebook by clicking "Save Version".

# <B>[5] Sample usage of utility scripts</B>

Sample usage of utility scripts is shown. [Utility scripts for visualizing dataset for "Happywhale - Whale and Dolphin Identification" competition](https://www.kaggle.com/code/acchiko/utility-functions-for-visualization-of-dataset) is loaded as example. Then several images are shown using utility functions defined in the utility scripts.

```python
# Imports utility scripts
import utility_functions_for_visualization_of_dataset as myutils
```

```python
# Imports required libs.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import os
```

```python
# Sets configurations.
num_images = 30
i_start = 0
i_end = i_start + num_images
num_cols = 4
```

```python
# Loads metadata.
path_to_train_metadata = "/kaggle/input/happy-whale-and-dolphin/train.csv"
metadata = pd.read_csv(path_to_train_metadata)
sliced_metadata = metadata.iloc[i_start:i_end]
```

```python
# Shows images.
path_to_dir_train_images = "/kaggle/input/happy-whale-and-dolphin/train_images"
titles = [row.image for row in sliced_metadata.itertuples()]
path_to_images = ["%s/%s" % (path_to_dir_train_images, row.image) for row in sliced_metadata.itertuples()]
images = myutils.getImages(path_to_images)
myutils.showImagesTile(titles, images, num_cols=num_cols)
```
