## Introduction

I built the notebook to randomly pick pictures of the species you're interested.

For example, there is one picture showing the **fin_whale** below.

You may choose the species you want to see and click the **Click me** button.

```python
# load dataset
data = pd.read_csv("../input/happy-whale-and-dolphin/train.csv")

# unique individual id
unique_individual_id = data['individual_id'].unique()

# choose species
# randomly pick one picture
# display it
def choose_species(num_pics,species,width,height,same_individual):
    if same_individual == True:
        dataset = data.loc[data['species'] == 'common_dolphin']
        unique_individual_id = dataset['individual_id'].unique()
        choose_id = random.choice(unique_individual_id)
        try:
            dataset = dataset.loc[dataset['individual_id'] == choose_id].sample(num_pics)
            pics_id = list(dataset['image'])
            individual_id = list(dataset['individual_id'])
            for i in range(num_pics):
                im = Image.open("../input/happy-whale-and-dolphin/train_images/" + str(pics_id[i]))
                resized_im = im.resize((width, height))
                print("individual_id: " + individual_id[i])
                display(resized_im)
        except ValueError:
            print("This individual doesn't have enough pics to show.")
            print("Please set a lower number of pictures.")
    else:
        dataset = data.loc[data['species'] == 'common_dolphin'].sample(num_pics)
        pics_id = list(dataset['image'])
        individual_id = list(dataset['individual_id'])
        for i in range(num_pics):
            im = Image.open("../input/happy-whale-and-dolphin/train_images/" + str(pics_id[i]))
            resized_im = im.resize((width, height))
            print("individual_id: " + individual_id[i])
            display(resized_im)
```

## Example
Here we can see two random pictures of one same fin_whale individual, either of the pic has width of 196 and height of 160. 

```python
choose_species(num_pics = 2,
               species = "fin_whale",
               width = 196,
               height = 160,
               same_individual = True)
```

## Choose the pics you are interested to see

**It is required to run the code from your own notebook, so you may need to fork this one.**

```python
# create dropdown menus to get input
unique_species = data['species'].unique()
output = Output()
start = Button(description="Click me")
start.style.button_color = 'lightblue'


num_widget = widgets.Dropdown(
    options=[1,2,3,4,5],
    value=list(list([1,2,3,4,5]))[0],
    description='Nr. of pics',
    disabled=False,
)

species_widget = widgets.Dropdown(
    options=list(unique_species),
    value=list(list(unique_species))[10],
    description='Species',
    disabled=False,
)

same_widget = widgets.Checkbox(
    value=False,
    description='Same individual',
    disabled=False,
    indent=False
)

width_widgets = widgets.IntSlider(
    value=144,
    min=0,
    max=360,
    step=5,
    description='width',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d',
)

height_widgets = widgets.IntSlider(
    value=128,
    min=0,
    max=360,
    step=5,
    description='height',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d',
)

def click_start(b):
    with output:
        clear_output()
        print("The species you choose is : " + species_widget.value)
        choose_species(num_widget.value,
                      species_widget.value,
                      width_widgets.value,
                      height_widgets.value,
                      same_widget.value)
        print("Click the button to get another one!")

start.on_click(click_start)

display(num_widget,
        species_widget,
        same_widget,
        width_widgets,
        height_widgets,
        start,
        output)


```

## Can you guess the species of this photo?

Let's play something for fun.

The program will randomly pick a picture to display and you can choose which species is the photo coming from.

```python
# choose one individual by random
species = data['species'].unique()
# list of wrong options
wrong_species = list(random.sample(list(species), 3))
sample = data.sample(1).iloc[0]
random_img = sample['image']
species_sample = sample['species']
pic_route = "../input/happy-whale-and-dolphin/train_images/" + str(random_img)
im = Image.open(pic_route)
resized_im_guess = im.resize((144, 144))
print("Do you know which species it is?")

# display widgets
check = Button(description="Check!")
check.style.button_color = 'lightblue'

species_choice = widgets.RadioButtons(
    options=[*wrong_species, species_sample],
    description='Species is ',
    disabled=False
)

output_2 = Output()




def click_check(b):
    with output_2:
        clear_output()
        if species_choice.value == species_sample:
            print("Bingo! You are right!")
        else:
            print("Hmmm... the species is " + str(species_sample) + " actually.")

        
check.on_click(click_check)

display(resized_im_guess,
        species_choice,
        check,
        output_2)
```
