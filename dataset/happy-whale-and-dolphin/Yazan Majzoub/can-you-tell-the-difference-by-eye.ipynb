```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

# **Show two pictures of each type to compare the types and know the common**

```python
data=pd.read_csv('../input/happy-whale-and-dolphin/train.csv')
data
```

```python
unique=data['species'].unique()
```

```python
for d in unique :
    a=data[data['species']==f'{d}'][0:1]['image'].values[0]
    b=data[data['species']==f'{d}'][1:2]['image'].values[0]
    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot()
    plt.imshow(plt.imread(f'../input/happy-whale-and-dolphin/train_images/{a}'))
    ax.set_title(f'{d}')
    
    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot()
    plt.imshow(plt.imread(f'../input/happy-whale-and-dolphin/train_images/{b}'))
    ax.set_title(f'{d}')
    
    
```
