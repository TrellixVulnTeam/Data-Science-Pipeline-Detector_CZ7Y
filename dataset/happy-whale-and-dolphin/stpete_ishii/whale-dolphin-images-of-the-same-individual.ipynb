# Images of the same individual id

```python
dir0='../input/happy-whale-and-dolphin/train_images'
```

```python
id0='c995c043c353'
imgs=train[train['individual_id']==id0]['image'].tolist()
species=train[train['individual_id']==id0]['species'].tolist()[0]
print('Individual_ID:',id0, ', Species:',species)

img_paths=[]
for item in imgs:
    img_paths+=[os.path.join(dir0,item)]

fig, axs = plt.subplots(4,5, figsize=(18,12))
for j in range(20):
    r=j//5
    c=j%5
    img0=cv2.imread(img_paths[j])
    axs[r][c].imshow(cv2.cvtColor(img0, cv2.COLOR_BGR2RGB))
    axs[r][c].axis("off")
plt.show()
```

```python
# right angle
id0='4b00fe572063'
imgs=train[train['individual_id']==id0]['image'].tolist()
species=train[train['individual_id']==id0]['species'].tolist()[0]
print('Individual_ID:',id0, ', Species:',species)

img_paths=[]
for item in imgs:
    img_paths+=[os.path.join(dir0,item)]

fig, axs = plt.subplots(4,5, figsize=(18,12))
for j in range(20):
    r=j//5
    c=j%5
    img0=cv2.imread(img_paths[j])
    a=img0.shape[0]
    b=img0.shape[1]
    axs[r][c].imshow(cv2.cvtColor(img0[:,(b-a):,:], cv2.COLOR_BGR2RGB))
    axs[r][c].axis("off")
plt.show()
```

```python
# left angle
id0='4b00fe572063'
imgs=train[train['individual_id']==id0]['image'].tolist()
species=train[train['individual_id']==id0]['species'].tolist()[0]
print('Individual_ID:',id0, ', Species:',species)

img_paths=[]
for item in imgs:
    img_paths+=[os.path.join(dir0,item)]

fig, axs = plt.subplots(4,5, figsize=(18,12))
for j in range(20):
    r=j//5
    c=j%5
    img0=cv2.imread(img_paths[j])
    a=img0.shape[0]
    b=img0.shape[1]
    axs[r][c].imshow(cv2.cvtColor(img0[:,0:a,:], cv2.COLOR_BGR2RGB))
    axs[r][c].axis("off")
plt.show()
```

```python
# enlarge center
id0='4b00fe572063'
imgs=train[train['individual_id']==id0]['image'].tolist()
species=train[train['individual_id']==id0]['species'].tolist()[0]
print('Individual_ID:',id0, ', Species:',species)

img_paths=[]
for item in imgs:
    img_paths+=[os.path.join(dir0,item)]

fig, axs = plt.subplots(4,5, figsize=(18,12))
for j in range(20):
    r=j//5
    c=j%5
    img0=cv2.imread(img_paths[j])
    a=img0.shape[0]//4
    b=img0.shape[1]//4
    axs[r][c].imshow(cv2.cvtColor(img0[a:a*3,b:b*3,:], cv2.COLOR_BGR2RGB))
    axs[r][c].axis("off")
plt.show()
```
