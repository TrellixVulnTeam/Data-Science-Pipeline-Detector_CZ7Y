```python
colorizer = get_image_colorizer(artistic=True)
```

```python
!mkdir train_imgs
train_imgs = os.listdir("../input/happy-whale-and-dolphin/train_images")
train_imgs = np.array(train_imgs)
bw_img = np.array([])
from tqdm.notebook import tqdm 
for i in tqdm(train_imgs):
    img = plt.imread(f"../input/happy-whale-and-dolphin/train_images/{i}")
    if len(img.shape)!=3 :
        image= colorizer.get_transformed_image(path = f"../input/happy-whale-and-dolphin/train_images/{i}", render_factor = 40 , watermarked = False)
        image.save("./train_imgs/"+i)
        bw_img = np.append(bw_img,f"../input/happy-whale-and-dolphin/train_images/{i}")
```

```python
np.save('train_bw', bw_img)
```

```python
!mkdir test_imgs
test_imgs = np.array(os.listdir("../input/happy-whale-and-dolphin/test_images"))

bw_img_test = np.array([])
from tqdm.notebook import tqdm 
for i in tqdm(test_imgs):
    img = plt.imread(f"../input/happy-whale-and-dolphin/test_images/{i}")
    if len(img.shape)!=3 :
        image= colorizer.get_transformed_image(path = f"../input/happy-whale-and-dolphin/test_images/{i}", render_factor = 40 , watermarked = False)
        image.save("./test_imgs/"+i)
        bw_img_test= np.append(bw_img_test,f"../input/happy-whale-and-dolphin/test_images/{i}")
```

```python
np.save('test_bw', bw_img_test)
```
