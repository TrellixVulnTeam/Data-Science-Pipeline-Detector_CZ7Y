```python
IMAGE_SIZE = 75

X = []
y = []

for i in tqdm(range(len(train_data))):
    row = list(train_data.iloc[i])
    img = cv2.imread(os.path.join('../input/happy-whale-and-dolphin/train_images', row[0]))
    img = cv2.resize(img, (IMAGE_SIZE, IMAGE_SIZE))
    y.append(row[-1])
    X.append(img)
```

```python
pickle_out = open(f"X_{IMAGE_SIZE}_arr.pickle", "wb")
pickle.dump(X, pickle_out)
pickle_out.close()

pickle_out = open(f"y_{IMAGE_SIZE}_arr.pickle", "wb")
pickle.dump(y, pickle_out)
pickle_out.close()
```
