```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Data exploration

```python
train_df.info()
```

```python
print("Number of unique species:",train_df['species'].nunique())
print("Species names:" ,train_df["species"].unique())
```

```python
specie_distribution = train_df['species'].value_counts()
specie_distribution
```

```python
specie_distribution.plot(kind="bar")
plt.show()
```

```python
print("Number of unique IDs:",train_df['individual_id'].nunique())
print("Average number of unique IDs:",len(train_df) / train_df['individual_id'].nunique())
```

```python
id_count = train_df['individual_id'].value_counts()
id_count
```

```python
id_count.plot()
plt.show()
```

```python
(id_count == 1).value_counts()
```

About 60% of IDs have data for only one case.

```python
small_data_count = (id_count[id_count <= 8]).value_counts()
small_data_count
```

```python
small_data_count.plot(kind="bar")
plt.show()
```
