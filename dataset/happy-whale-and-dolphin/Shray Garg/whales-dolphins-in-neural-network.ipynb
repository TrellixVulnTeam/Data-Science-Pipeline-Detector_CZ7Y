> ## As the dataset is so huge, let's specify directory paths and get Image paths

```python
train_directory = '../input/happy-whale-and-dolphin/train_images'
test_directory = '../input/happy-whale-and-dolphin/test_images'

def getImagePaths(path):
    images = []
    for directoryname, _, filename in os.walk(path):
        for fname in filename:
            fullpath = os.path.join(directoryname, fname)
            images.append(fullpath)
    return images

trainimages_path = getImagePaths(train_directory)
testimages_path = getImagePaths(test_directory)
```
