```python
import shutil
import numpy as np
import pandas as pd
import os

import cv2
from tqdm import tqdm

from joblib import Parallel, delayed
```

```python
train = pd.read_csv("../input/happy-whale-and-dolphin/train.csv")
```

```python
NEW_SIZE = (128, 128) # (height, width)
INTERPOLATION = cv2.INTER_CUBIC
```

```python
def load_and_write(path, image, output_path):
    img = cv2.imread(path)
    img = cv2.resize(img, NEW_SIZE[::-1], interpolation=INTERPOLATION)
    cv2.imwrite(os.path.join(output_path , image), img)
    
def create_resized_images(image_path, image_output_path):
    image_list = os.listdir(image_path)
    Parallel(n_jobs=20)(
        delayed(load_and_write)(
            os.path.join(image_path, image),
            image,
            image_output_path
        ) for image in tqdm(image_list)
    )
```

```python
def resize_images(input_folder, output_folder):
    os.makedirs(output_folder)
    create_resized_images(input_folder, output_folder)
    shutil.make_archive(f"{output_folder}_zip", 'zip', output_folder)
```

```python
train_image_path = "../input/happy-whale-and-dolphin/train_images"
train_image_output_path = f"/kaggle/working/train_images_{NEW_SIZE[0]}"

resize_images(train_image_path, train_image_output_path)
```

```python
test_image_path = "../input/happy-whale-and-dolphin/test_images"
test_image_output_path = f"/kaggle/working/test_images_{NEW_SIZE[0]}"

resize_images(test_image_path, test_image_output_path)
```
