````python
print(f"\n... ACCELERATOR SETUP STARTING ...\n")

# Detect hardware, return appropriate distribution strategy
try:
    # TPU detection. No parameters necessary if TPU_NAME environment variable is set. On Kaggle this is always the case.
    TPU = tf.distribute.cluster_resolver.TPUClusterResolver()  
except ValueError:
    TPU = None

if TPU:
    print(f"\n... RUNNING ON TPU - {TPU.master()}...")
    tf.config.experimental_connect_to_cluster(TPU)
    tf.tpu.experimental.initialize_tpu_system(TPU)
    strategy = tf.distribute.experimental.TPUStrategy(TPU)
else:
    print(f"\n... RUNNING ON CPU/GPU ...")
    # Yield the default distribution strategy in Tensorflow
    #   --> Works on CPU and single GPU.
    strategy = tf.distribute.get_strategy() 

# What Is a Replica?
#    --> A single Cloud TPU device consists of FOUR chips, each of which has TWO TPU cores. 
#    --> Therefore, for efficient utilization of Cloud TPU, a program should make use of each of the EIGHT (4x2) cores. 
#    --> Each replica is essentially a copy of the training graph that is run on each core and 
#        trains a mini-batch containing 1/8th of the overall batch size
N_REPLICAS = strategy.num_replicas_in_sync
    
print(f"... # OF REPLICAS: {N_REPLICAS} ...\n")

print(f"\n... ACCELERATOR SETUP COMPLTED ...\n")
````

---


**XLA** (Accelerated Linear Algebra) is a domain-specific compiler for linear algebra that can accelerate TensorFlow models with potentially no source code changes. **The results are improvements in speed and memory usage**.

<br>

When a TensorFlow program is run, all of the operations are executed individually by the TensorFlow executor. Each TensorFlow operation has a precompiled GPU/TPU kernel implementation that the executor dispatches to.

XLA provides us with an alternative mode of running models: it compiles the TensorFlow graph into a sequence of computation kernels generated specifically for the given model. Because these kernels are unique to the model, they can exploit model-specific information for optimization.<br><br>

<div class="alert alert-block alert-danger" style="margin: 2em; line-height: 1.7em; font-family: Verdana;">
    <b style="font-size: 16px;">üõë &nbsp; WARNING:</b><br><br>- XLA can not currently compile functions where dimensions are not inferrable: that is, if it's not possible to infer the dimensions of all tensors without running the entire computation<br>
</div>

````python
print("\n... BASIC DATA SETUP STARTING ...\n\n")

K_FOLDS = N_FOLDS = 10 # More later...
ALL_TFRECORDS = tf.io.gfile.glob(os.path.join(DATA_DIR, "**/*.tfrec"))
tfrecord_map = {
    "train":{f"fold_{i}":[_tfrec for _tfrec in ALL_TFRECORDS if f"train_fold_{i}" in _tfrec] for i in range(1, N_FOLDS+1)},
    "test":[_tfrec for _tfrec in ALL_TFRECORDS if "test__" in _tfrec]
}

train_df = pd.read_csv("../input/happy-whale-and-dolphin/train.csv")
FIX_NAME_MAPPING = {"bottlenose_dolpin":"bottlenose_dolphin", 
                    "kiler_whale":"killer_whale",
                    "pilot_whale":"short_finned_pilot_whale",
                    "globis":"short_finned_pilot_whale"}
train_df["species"] = train_df["species"].apply(lambda x: x if x not in FIX_NAME_MAPPING.keys() else FIX_NAME_MAPPING[x])

all_species = sorted(train_df.species.unique().tolist())
species_int2str_lbl_map = {i:_s for i,_s in enumerate(all_species)}
species_str2int_lbl_map = {v:k for k,v in species_int2str_lbl_map.items()}
N_SPECIES = len(all_species)

all_individuals = sorted(train_df.individual_id.unique().tolist())
ind_int2str_lbl_map = {i:_s for i,_s in enumerate(all_individuals)}
ind_str2int_lbl_map = {v:k for k,v in ind_int2str_lbl_map.items()}
N_INDIVIDUALS = len(all_individuals)

train_df["ind_sparse_lbl"] = train_df["individual_id"].map(ind_str2int_lbl_map)
train_df["species_sparse_lbl"] = train_df["species"].map(species_str2int_lbl_map)

ss_df = pd.read_csv("../input/happy-whale-and-dolphin/sample_submission.csv")
ss_df["img_path"] = "../input/happy-whale-and-dolphin/test_images/"+ss_df.image

sub_train_dfs = [pd.read_csv(f"/kaggle/input/happywhaletfrecords512x880/train_df_fold_{i}.csv") for i in range(1,K_FOLDS+1)]
N_TEST = len(ss_df)

print("\n\n... BASIC DATA SETUP FINISHING ...\n")
````

````python
cls_counts = {ind_str2int_lbl_map[k]:v for k,v in train_df.individual_id.value_counts().items()} 
cls_min_count = min(list(cls_counts.values()))

# This will result in very small weighting for
# 'common' individuals and a weighting of
# 1.00 for the rarest individuals (probably only 1 image)
orig_cls_wt = {k:cls_min_count/v for k,v in cls_counts.items()}

# One approach is to raise all the weights to a small value
# between 0 and 1. 
# The closer the value is to 0 the closer the weights will be to 1. 
# The closer the value is to 1 the closer to the original values. 
flat_pwr = 0.1
flat_cls_wt_1 = {k:(cls_min_count/v)**flat_pwr for k,v in cls_counts.items()}

# Another step would be to use a slightly higher value for the flat_pwr
# term, and clip the weights to a minimum to prevent certain very common
# individuals from skewing the class balance
flat_pwr = 0.25
clip_val = 0.25
flat_cls_wt_2 = {k:max(clip_val, (cls_min_count/v)**flat_pwr) for k,v in cls_counts.items()}

fig = px.histogram(orig_cls_wt.values(), labels={"value": "class weight"}, title="<b>Original Weight Distribution</b>", range_x=[0,1], log_y=True)
fig.update_layout(showlegend=False) 
fig.show()

fig = px.histogram(flat_cls_wt_1.values(), labels={"value": "class weight"}, title="<b>Shifted Weight Distribution</b>", range_x=[0,1], log_y=True)
fig.update_layout(showlegend=False) 
fig.show()

fig = px.histogram(flat_cls_wt_2.values(), labels={"value": "class weight"}, title="<b>Shifted and Clipped Weight Distribution</b>", range_x=[0,1], log_y=True)
fig.update_layout(showlegend=False) 
fig.show()
````

````python
def flatten_l_o_l(nested_list):
    """ Flatten a list of lists """
    return [item for sublist in nested_list for item in sublist]


def decode_image(image_data, resize_to=(512,880,3)):
    """ Function to decode the tf.string containing image information 
    
    
    Args:
        image_data (tf.string): String containing encoded image data from tf.Example
        resize_to (tuple, optional): Size that we will reshape the tensor to (required for TPU)
        to_norm_val (int, optional): The value that will allow us to normalize the tensor to 0-1
    
    Returns:
        Tensor containing the resized single-channel image in the appropriate dtype
    """
    image = tf.image.decode_png(image_data, channels=3, dtype=tf.uint8)
    image = tf.reshape(image, resize_to)
    return tf.cast(image, tf.float32)


def decode_example(serialized_example, is_test=False, return_ind_id=False, return_img_id=False, img_shape=(512,880,3)):
    """ Parses a set of features and label from the given `serialized_example`.
        
        It is used as a map function for `dataset.map`

    Args:
        serialized_example (tf.Example): A serialized example containing the
            following features:
                ‚Äì 'image'
                ‚Äì 'image_id'
                ‚Äì 'label'
        is_test (bool, optional): Whether to allow for the label feature
        
    Returns:
        A decoded tf.data.Dataset object representing the tfrecord dataset
    """
    feature_dict = {
        'image': tf.io.FixedLenFeature(shape=[], dtype=tf.string, default_value=''),
    }
    
    if not is_test:
        feature_dict["label"] = tf.io.FixedLenFeature(shape=[], dtype=tf.int64, default_value=0)
        feature_dict["ind_id"] = tf.io.FixedLenFeature(shape=[], dtype=tf.int64, default_value=0)
    feature_dict["image_id"] = tf.io.FixedLenFeature(shape=[], dtype=tf.string, default_value='')
    
    # Define a parser
    features = tf.io.parse_single_example(serialized_example, features=feature_dict)
    
    # Decode the tf.string
    image = decode_image(features['image'], resize_to=img_shape)
    
    # Figure out the correct information to return
    if is_test:
        image_id = features["image_id"] 
        return image, image_id
    else:
        label = features["label"]
        if (return_ind_id or return_img_id):
            ind_id = features["ind_id"]
            image_id = features["image_id"]
            return image, label, ind_id, image_id
        else:
            return image, label
    
DEFAULT_AUG_CONFIG = dict(
    random_hue=dict(max_delta=0.01,), random_brightness=dict(max_delta=0.1,),
    random_saturation=dict(lower=0.75, upper=1.25), random_contrast=dict(lower=0.875, upper=1.125)
)
def augment(img_batch, cutout_central_fraction=0.95, cutout_ratio=26, aug_config=DEFAULT_AUG_CONFIG):
    
    # SEEDING & KERNEL INIT
    img_batch = tf.cast(img_batch, tf.float32)
    
    # Random Flip
    img_batch = tf.image.random_flip_left_right(img_batch,)
    
    # Central Crop
    K = tf.random.uniform((1,), minval=0, maxval=4, dtype=tf.dtypes.int32)[0]    
    img_batch = tf.cond(K==1, lambda: tf.image.resize(
        tf.image.central_crop(img_batch, central_fraction=cutout_central_fraction), 
        (IMG_SHAPE[0], IMG_SHAPE[1])
    ), lambda: img_batch)
    
    # Random Cutout (black squares)
    img_batch = tfa.image.random_cutout(img_batch, mask_size=(2*(IMG_SHAPE[0]//cutout_ratio),2*(IMG_SHAPE[0]//cutout_ratio)), constant_values=0)
    
    # Visual Augmentations
    img_batch = tf.image.random_hue(img_batch, **aug_config["random_hue"])
    img_batch = tf.image.random_saturation(img_batch, **aug_config["random_saturation"])
    img_batch = tf.image.random_contrast(img_batch, **aug_config["random_contrast"])
    img_batch = tf.image.random_brightness(img_batch, **aug_config["random_brightness"])
    
    return tf.cast(img_batch, tf.float32)


# UNNECESSARY NOW
# # build a lookup table so we can do a dictionary lookup in graph mode
# s2i_table = tf.lookup.StaticHashTable(
#     initializer=tf.lookup.KeyValueTensorInitializer(
#         keys=tf.constant(list(ind_str2int_lbl_map.keys())),
#         values=tf.constant(list(ind_str2int_lbl_map.values())),
#     ),
#     default_value=tf.constant(-1),
#     name="str2int_mapping"
# )
````

````python
class ArcMarginProduct(tf.keras.layers.Layer):
    """
    KAGGLE REFERENCE  
        --> https://www.kaggle.com/ragnar123/unsupervised-baseline-arcface
        
    Implements large margin arc distance.

    Reference:
        https://arxiv.org/pdf/1801.07698.pdf
        https://github.com/lyakaap/Landmark2019-1st-and-3rd-Place-Solution/
            blob/master/src/modeling/metric_learning.py
    """
    def __init__(self, n_classes, s=30, m=0.3,
                 easy_margin=False, ls_eps=0.0,  **kwargs):
        super(ArcMarginProduct, self).__init__(**kwargs)

        self.n_classes = n_classes
        self.s, self.m = s, m
        self.ls_eps = ls_eps
        self.easy_margin = easy_margin
        self.sin_m, self.cos_m = tf.math.sin(m), tf.math.cos(m)
        self.mm, self.th = tf.math.sin(math.pi-m)*m, tf.math.cos(math.pi-m)

    def get_config(self):
        config = super().get_config().copy()
        config.update({
            'n_classes': self.n_classes, 's': self.s, 'm': self.m,
            'ls_eps': self.ls_eps, 'easy_margin': self.easy_margin,
        })
        return config

    def build(self, input_shape):
        super(ArcMarginProduct, self).build(input_shape[0])

        self.W = self.add_weight(
            name='W', shape=(int(input_shape[0][-1]), self.n_classes),
            initializer='glorot_uniform', dtype='float32', 
            trainable=True, regularizer=None)

    def call(self, inputs):
        X, y = inputs
        y = tf.cast(y, dtype=tf.int32)
        cosine = tf.matmul(
            tf.math.l2_normalize(X, axis=1),
            tf.math.l2_normalize(self.W, axis=0)
        )
        sine = tf.math.sqrt(1.0 - tf.math.pow(cosine, 2))
        phi = cosine * self.cos_m - sine * self.sin_m
        
        if self.easy_margin:
            phi = tf.where(cosine > 0, phi, cosine)
        else:
            phi = tf.where(cosine > self.th, phi, cosine - self.mm)
        
        one_hot = tf.cast(
            tf.one_hot(y, depth=self.n_classes),
            dtype=cosine.dtype
        )
        
        if self.ls_eps > 0:
            one_hot = (1 - self.ls_eps) * one_hot + self.ls_eps / self.n_classes

        output = (one_hot * phi) + ((1.0 - one_hot) * cosine)
        output *= self.s
        
        return output
````

````python
def get_train_val_split(ds_map, val_fold="1"):
    # Get shuffled training folds as tf.data.Dataset
    train_folds = [
        str(x) for x \
        in random.sample(range(1, 11), 10) \
        if str(x)!=str(val_fold)
    ]
    
    print(train_folds)
    print(val_fold)
    _train_ds = ds_map["train"][f"fold_{train_folds[0]}"]
    for _fold_num in train_folds[1:]:
        _train_ds = _train_ds.concatenate(ds_map["train"][f"fold_{_fold_num}"])
        
    # Get validation dataset as tf.data.Dataset
    _val_ds = ds_map["train"][f"fold_{val_fold}"]
    return _train_ds, _val_ds

def return_padded_ds(_ds, ds_length, is_test=False):
    def get_n_pad(_ds_length):
        return BATCH_SIZE-_ds_length%BATCH_SIZE
    # Get amount to pad
    n_pad = get_n_pad(ds_length)
    
    if is_test:
        _pad_ds = tf.data.Dataset.from_tensor_slices((
                    tf.zeros((n_pad, *INPUT_SHAPE), dtype=tf.float32),    # Fake Images
                    tf.constant(["000000000000",]*n_pad, dtype=tf.string) # Fake Image ID
        ))
        _ds = _ds.concatenate(_pad_ds) # update in place
    else:
        _pad_ds = tf.data.Dataset.from_tensor_slices((
                tf.zeros((n_pad, *INPUT_SHAPE), dtype=tf.float32),    # Fake Images
                tf.zeros((n_pad,), dtype=tf.int64),                   # Fake Label
                tf.zeros((n_pad,), dtype=tf.int64),                   # Fake Individual
                tf.constant(["000000000000",]*n_pad, dtype=tf.string) # Fake Image ID
        ))
        _ds = _ds.concatenate(_pad_ds) # make new dataset
    return _ds

def get_lr_callback(do_plot=True):
    """
    https://www.kaggle.com/ragnar123/shopee-efficientnetb3-arcmarginproduct
    """
    lr_start   = 0.000001
    lr_max     = 0.000005*BATCH_SIZE
    lr_min     = 0.000001
    lr_ramp_ep = 5
    lr_sus_ep  = 2
    lr_decay   = 0.9
   
    def lrfn(epoch):
        if epoch < lr_ramp_ep:
            lr = (lr_max - lr_start) / lr_ramp_ep * epoch + lr_start   
        elif epoch < lr_ramp_ep + lr_sus_ep:
            lr = lr_max    
        else:
            lr = (lr_max - lr_min) * lr_decay**(epoch - lr_ramp_ep - lr_sus_ep) + lr_min    
        return lr
    
    lr_callback = tf.keras.callbacks.LearningRateScheduler(lrfn, verbose = True)
    
    fig = px.scatter(x=[x for x in range(N_EPOCHS)], y=[lrfn(x) for x in range(N_EPOCHS)], title="Learning Rate Schedule", labels={"x":"<b>Epoch Number</b>", "y":"<b>Learning Rate</b>"})
    fig.show()
    return lr_callback

def plot_history(history, fold_num="1"):
    fig = px.line(history.history, 
                  x=range(len(history.history["loss"])), 
                  y=["loss", "val_loss"],
                  labels={"value":"Loss (log-axis)", "x":"Epoch #"},
                  title=f"<b>FOLD {fold_num} MODEL - LOSS</b>", log_y=True
                  )
    fig.show()

    fig = px.line(history.history, 
                  x=range(len(history.history["acc"])), 
                  y=["acc", "val_acc"],
                  labels={"value":"Accuracy (log-axis)", "x":"Epoch #"},
                  title=f"<b>FOLD {fold_num} MODEL - ACCURACY</b>", log_y=True)
    fig.show()

class GeMPoolingLayer(tf.keras.layers.Layer):
    def __init__(self, p=3., train_p=False):
        super().__init__()
        if train_p:
            self.p = tf.Variable(p, dtype=tf.float32)
        else:
            self.p = p
        self.eps = 1e-6

    def call(self, inputs: tf.Tensor, **kwargs):
        inputs = tf.clip_by_value(inputs, clip_value_min=1e-6, clip_value_max=tf.reduce_max(inputs))
        inputs = tf.pow(inputs, self.p)
        inputs = tf.reduce_mean(inputs, axis=[1, 2], keepdims=False)
        inputs = tf.pow(inputs, 1./self.p)
        return inputs
    
def get_model(backbone, input_shape=INPUT_SHAPE, use_GeM=False, freeze_bn=False, n_dense=512, dropout=0.025, n_classes=N_INDIVIDUALS):
    """ TBD """
    # Define the ArcMargin head layer
    MarginLayer = ArcMarginProduct(
        n_classes=n_classes, name='head/arc_margin', dtype='float32',
    )
    GeMLayer = GeMPoolingLayer()
    
    # Define inputs
    #   input_1 --> image
    #   input_2 --> label
    _input_1 = tf.keras.layers.Input(shape = INPUT_SHAPE, name = 'input_1')
    _input_2 = tf.keras.layers.Input(shape = (),          name = 'input_2')
    
    x = backbone(_input_1)
    
    if use_GeM:
        x = GeMLayer(x)
    
    x = tf.keras.layers.Dropout(dropout)(x)
    x = tf.keras.layers.Dense(n_dense, name="embedding_out")(x) # no activation
    x = MarginLayer([x, _input_2])
    _output = tf.keras.layers.Softmax(dtype='float32')(x)    
    return tf.keras.models.Model(inputs = [_input_1, _input_2], outputs = [_output])

def get_embed_model(model, bb_layer_name="embedding_out"):
    _inputs = model.input[0]
    _outputs = model.get_layer(bb_layer_name).output
    return tf.keras.Model(inputs=_inputs, outputs=_outputs)

def get_embedding(_ds, _emodel, clip_pad_at, style="train", to_disk=True, fold_num="1"):
    print(f"\n\n\n... GENERATING {style.upper()} EMBEDDINGS ...\n\n")
    _embeds, _labels, _individual_ids, _image_ids = [], [], [], []
    
    if style=="test":
        for img_batch, img_id_batch in tqdm(_ds, total=clip_pad_at//BATCH_SIZE):
            pred_batch = _emodel.predict(img_batch)
            _embeds.append(pred_batch)
            _image_ids.append(img_id_batch)    
        _embeds = tf.concat(_embeds, axis=0).numpy()[:clip_pad_at]
        _image_ids = tf.concat(_image_ids, axis=0).numpy()[:clip_pad_at]
    else:
        for img_batch, lbl_batch, ind_id_batch, img_id_batch in tqdm(_ds, total=clip_pad_at//BATCH_SIZE):
            pred_batch = _emodel.predict(img_batch)
            _embeds.append(pred_batch)
            _image_ids.append(img_id_batch)    
            _labels.append(lbl_batch)
            _individual_ids.append(ind_id_batch)
        _embeds = tf.concat(_embeds, axis=0).numpy()[:clip_pad_at]
        _image_ids = tf.concat(_image_ids, axis=0).numpy()[:clip_pad_at]
        _individual_ids = tf.concat(_individual_ids, axis=0).numpy()[:clip_pad_at]
        _labels = tf.concat(_labels, axis=0).numpy()[:clip_pad_at]
    
    if to_disk:
        np.save(f"fold_{fold_num}__{style}_image_embeds", _embeds)
        np.save(f"fold_{fold_num}__{style}_image_ids", _image_ids)
        if style!="test":
            np.save(f"fold_{fold_num}__{style}_individual_ids", _individual_ids)
            np.save(f"fold_{fold_num}__{style}_image_labels", _labels)
    else:
        raise NotImplementedError()
````

````python
# Step 0: Initialize Histories
histories = []

# Step 1: Add Padding to Test Dataset
test_ds = return_padded_ds(tfds_map["test"], ds_length=len(ss_df), is_test=True)

# Step 2: Preprocess Test Dataset
test_ds = test_ds.batch(BATCH_SIZE, drop_remainder=True) \
                 .map(lambda x,y: (x/255., y)) \
                 .prefetch(tf.data.AUTOTUNE)

# Step 3: Start Looping
for _k in range(7, K_FOLDS+1):
    print(f"\n\n\n\n\n... STARTING TRAINING FOR FOLD {_k} ...\n\n\n")
    # Step 4: Get train and val split based on fold
    #            --> Train = 9 folds
    #            --> Val   = 1 fold
    train_ds, val_ds = get_train_val_split(tfds_map, val_fold=str(_k))
    
    # Step 5: Get a padded version of the datasets
    N_TRAIN = sum([len(x) for i, x in enumerate(sub_train_dfs) if i!=(_k-1)])
    N_VAL = len(sub_train_dfs[_k-1])
    padded_train_ds = return_padded_ds(train_ds, ds_length=N_TRAIN)
    padded_val_ds   = return_padded_ds(val_ds, ds_length=N_VAL)
    
    # Step 6: Preprocess the datasets
    train_ds = train_ds.shuffle(BUFFER_SIZE) \
                       .batch(BATCH_SIZE, drop_remainder=True) \
                       .map(lambda x,y,z,a: (augment(x)/255., tf.cast(y, tf.uint8), tf.cast(z, tf.int32), a), num_parallel_calls=tf.data.AUTOTUNE) \
                       .prefetch(tf.data.AUTOTUNE)
    
    # No augmentation for validation or padded datasets
    # We will still shuffle the validation dataset during training 
    # so we can see performance on all examples (as some will be dropped each epoch)
    val_ds = val_ds.shuffle(BUFFER_SIZE) \
                   .batch(BATCH_SIZE, drop_remainder=True) \
                   .map(lambda x,y,z,a: (x/255., tf.cast(y, tf.uint8), tf.cast(z, tf.int32), a), num_parallel_calls=tf.data.AUTOTUNE) \
                   .prefetch(tf.data.AUTOTUNE)
    padded_train_ds = padded_train_ds.batch(BATCH_SIZE, drop_remainder=True) \
                                     .map(lambda x,y,z,a: (x/255., tf.cast(y, tf.uint8), tf.cast(z, tf.int32), a), num_parallel_calls=tf.data.AUTOTUNE) \
                                     .prefetch(tf.data.AUTOTUNE)
    padded_val_ds = padded_val_ds.batch(BATCH_SIZE, drop_remainder=True) \
                                 .map(lambda x,y,z,a: (x/255., tf.cast(y, tf.uint8), tf.cast(z, tf.int32), a), num_parallel_calls=tf.data.AUTOTUNE) \
                                 .prefetch(tf.data.AUTOTUNE)

    # Step 7: Load the model within the strategy, compile and train
    with strategy.scope():
        if "v2" in MODEL_NAME:
            model = get_model(tfhub.KerasLayer(TFHUB_PATH, trainable=True, load_options=load_locally), n_dense=N_DENSE, dropout=DROPOUT, n_classes=N_INDIVIDUALS, use_GeM=False)
        else:
            model = get_model(efn.EfficientNetB6(include_top=False, weights='noisy-student'), n_dense=N_DENSE, dropout=DROPOUT, n_classes=N_INDIVIDUALS, use_GeM=USE_GEM)
        if _k==1:
            print(model.summary())
        
        # ######################################################################
        # Learning Rate Stuff
        # ######################################################################
        steps_per_epoch = N_TRAIN//BATCH_SIZE
        clr = tfa.optimizers.CyclicalLearningRate(initial_learning_rate=INIT_LR,
                                          maximal_learning_rate=MAX_LR,
                                          scale_fn=lambda x: 1/(2.5**(x-1)),
                                          step_size=4*steps_per_epoch)
        step = np.arange(0, N_EPOCHS*steps_per_epoch)
        lr = clr(step)
        fig = px.scatter(x=step, y=list(lr.numpy()), 
                         title="<b>Cyclical Learning Rate Schedule</b>",
                         labels={"x":"<b>Step</b>", "y":"<b>Learning Rate</b>"})
        fig.show()
        # ######################################################################
        
        model.compile(optimizer=tf.keras.optimizers.Adam(clr),
                     loss=tf.keras.losses.SparseCategoricalCrossentropy(),
                     metrics=["acc", tf.keras.metrics.SparseTopKCategoricalAccuracy(k=5)])
        
        ckpt_cb = tf.keras.callbacks.ModelCheckpoint(f'fold_{_k}__{MODEL_NAME}__{IMG_SHAPE[0]}x{IMG_SHAPE[1]}', 
                                                     monitor='val_loss', save_best_only=True,
                                                     options=save_locally, mode='min')
        # lr_cb = get_lr_callback()
        
    # Step 8: Keep the history and plot - ADDED CLASS WEIGHTING
    history = model.fit(train_ds.map(lambda a,b,c,d: ((a,c), c)), 
                        validation_data=val_ds.map(lambda a,b,c,d: ((a,c), c)), 
                        epochs=N_EPOCHS, callbacks = [ckpt_cb,],
                        class_weight=flat_cls_wt_2)
    plot_history(history)
    histories.append(history)
    
    # Step 9: Create the embedding model
    with strategy.scope():
        emodel = get_embed_model(model)
        emodel.trainable=False
    
    # Step 10: Generate and save embeddings
    get_embedding(padded_train_ds, emodel, clip_pad_at=N_TRAIN, style="train", fold_num=str(_k))
    get_embedding(padded_val_ds, emodel, clip_pad_at=N_VAL, style="val", fold_num=str(_k))
    get_embedding(test_ds, emodel, clip_pad_at=N_TEST, style="test", fold_num=str(_k))
````
