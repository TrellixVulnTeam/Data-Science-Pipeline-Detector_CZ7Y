```python
first_name_df = pd.read_excel("../input/us-first-names-databasee/SSA_Names_DB.xlsx")
common_names = first_name_df["Name"]

surname_df = pd.read_csv("../input/us-first-names-databasee/Common_Surnames_Census_2000.csv")
surnames = surname_df.name.str.capitalize()

train_df = pd.read_csv("../input/happy-whale-and-dolphin/train.csv")
```

```python
individual_id_to_first_name = {_ind_id:_name for _ind_id,_name in zip(train_df.individual_id.unique(), common_names.tolist())}
species_to_last_name = {_species:_surname for _species,_surname in zip(train_df.species.unique(), surnames.tolist())}

new_individual_first_name_list = common_names.tolist()[15587:]

train_df["first_name"] = train_df["individual_id"].map(individual_id_to_first_name)
train_df["last_name"] = train_df["species"].map(species_to_last_name)

train_df
```

```python
import pickle
import json

with open(r"./new_individual_first_name_list.pickle", "wb") as output_file:
    pickle.dump(new_individual_first_name_list, output_file)

with open(r"./individual_id_to_first_name.pickle", "wb") as output_file:
    pickle.dump(individual_id_to_first_name, output_file)

with open(r"./species_to_last_name.pickle", "wb") as output_file:
    pickle.dump(species_to_last_name, output_file)
    
with open('./individual_id_to_first_name.json', 'w') as fp:
    json.dump(individual_id_to_first_name, fp,  indent=4)
    
with open('./species_to_last_name.json', 'w') as fp:
    json.dump(species_to_last_name, fp,  indent=4)
```
