```python
import numpy as np
import pandas as pd

TRAIN_PATH = "../input/happy-whale-and-dolphin/train.csv"
TARGET1 = "species"
TARGET2 = "individual_id"
```

```python
train = pd.read_csv(TRAIN_PATH)
```

```python
def getLabelCount(df,target):
    return [( labelValue,len(train.loc[df[target] == labelValue]) ) for labelValue in df[target].unique()]
```

# species count

```python
labelCount = getLabelCount(train,TARGET1)
sorted(labelCount,reverse=True,key=lambda item: item[1])
```

# individual_id count

```python
labelCount = getLabelCount(train,TARGET2)
sorted(labelCount,reverse=True,key=lambda item: item[1])
```
