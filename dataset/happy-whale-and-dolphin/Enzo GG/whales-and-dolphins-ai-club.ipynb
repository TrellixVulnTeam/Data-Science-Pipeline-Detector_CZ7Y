```python
model = keras.Sequential([
    layers.InputLayer(input_shape=(300, 300, 3)),
    layers.Conv2D(kernel_size=(3, 3), activation='relu', filters=1),
    layers.MaxPooling2D(2,2),
    #layers.Dropout(0.3),
    #layers.BatchNormalization(),
    layers.Conv2D(kernel_size=(3, 3), activation='relu', filters=1),
    layers.MaxPooling2D(2,2),
    layers.Flatten(),
    layers.Dropout(0.1),
    layers.BatchNormalization(),
    layers.Dense(1486, activation='softmax')
])
```

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),
    loss='categorical_crossentropy',
    metrics='accuracy',
)
```

```python
early_stopping = keras.callbacks.EarlyStopping(
    min_delta=0.001,
    patience=12,
    restore_best_weights=True,
)
```

```python
x_train, y_train_np, x_val, y_val_np = np.asarray(x_train), np.asarray(y_train), np.asarray(x_val), np.asarray(y_val)
```

```python
x_train.shape, y_train_np.shape, x_val.shape, y_val_np.shape
```

```python
x_train, x_val = x_train.reshape(1400,300,300,3), x_val.reshape(600,300,300,3)
```

```python
history = model.fit(
    x_train, y_train_np,
    validation_data=(x_val, y_val_np),
    batch_size=32,
    epochs=125,
    #callbacks=[early_stopping],
)
```

```python
x_train.shape, x_val.shape
```

```python
history_df = pd.DataFrame(history.history)

history_df.loc[:, ['loss', 'val_loss']].plot()
history_df.loc[:, ['accuracy', 'val_accuracy']].plot()
```
