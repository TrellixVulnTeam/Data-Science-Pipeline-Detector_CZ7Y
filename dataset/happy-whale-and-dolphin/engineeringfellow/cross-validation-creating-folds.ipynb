```python
import numpy as np
import pandas as pd

from sklearn import datasets
from sklearn import model_selection
```

```python
#Article on Cross Validation : https://dev.to/rohitgupta24/cross-validation-5gm3
```

```python
df = pd.read_csv("../input/happy-whale-and-dolphin/train.csv")
df.species.value_counts()
```

```python
#Data Cleaning
#Found out some duplicate labels

print("Number of unique species before fixing : ", df['species'].nunique())

df['species'].replace({
    'bottlenose_dolpin' : 'bottlenose_dolphin',
    'kiler_whale' : 'killer_whale',
},inplace =True)

print("Number of unique species after fixing : ", df['species'].nunique())


# df['class'] = df['species'].apply(lambda x: x.split('_')[-1])
df.head()
```

```python
'''visualizing  the data count'''


#visualization imports
#to have better idea, please visit : https://www.kaggle.com/awsaf49/happywhale-data-distribution#Dolphin-Species-6%EF%B8%8F%E2%83%A3

import plotly.express as px

data = df.species.value_counts().reset_index()
fig = px.bar(data, x='index', y='species', color='species',title='Species Count', text_auto=True)
fig.update_traces(textfont_size=12, textangle=0, textposition="outside", cliponaxis=False)
fig.show()
```

```python
#dataset is skewed, hence it is an ideal case for stratified Cross Validation



# we create a new column called kfold and fill it with -1
df["kfold"] = -1
# the next step is to randomize the rows of the data
df = df.sample(frac=1).reset_index(drop=True)
# fetch targets
y = df.species.values
# initiate the kfold class from model_selection module
kf = model_selection.StratifiedKFold(n_splits=10)
# fill the new kfold column
for f, (t_, v_) in enumerate(kf.split(X=df, y=y)):
    df.loc[v_, 'kfold'] = f
```

```python
# save the new csv with kfold column
df.to_csv("train_folds.csv", index=False)
```
