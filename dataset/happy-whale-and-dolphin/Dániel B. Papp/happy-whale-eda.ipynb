```python
train_dir = '/kaggle/input/happy-whale-and-dolphin/train_images'
test_dir = '/kaggle/input/happy-whale-and-dolphin/test_images'

# function to get image paths from train and test directory

def getImagePaths(path):
    image_names = []
    for dirname, _, filenames in os.walk(path):
        for filename in filenames:
            fullpath = os.path.join(dirname, filename)
            image_names.append(fullpath)
    return image_names

train_images_path = getImagePaths(train_dir)
test_images_path = getImagePaths(test_dir)
```

```python
def frequency(df, col, freq):
    n = 10
    if freq == "Most":
        return df[col].value_counts()[:n].index.tolist()
    elif freq == "Least":
        return df[col].value_counts()[-n:].index.tolist()
    
most_freq_species = frequency(train_df,"species", "Most")
least_freq_species = frequency(train_df,"species", "Least")
most_freq_ID = frequency(train_df,"individual_id", "Most")
least_freq_ID = frequency(train_df,"individual_id", "Least")
```

```python
def path(df,group,group_type):
    PATH = "/kaggle/input/happy-whale-and-dolphin/train_images"
    
    #species
    if group_type=='sp':
        z = df['image'][df['species']==group].values 
    
    #ID
    if group_type=='id':
        z = df['image'][df['individual_id']==group].values 
   
    image_names = []
    for filename in z:
        fullpath = os.path.join(PATH, filename)
        image_names.append(fullpath)
    return image_names
```

```python
def display_multiple_img(images_paths, rows, cols,title):
    
    figure, ax = plt.subplots(nrows=rows,ncols=cols,figsize=(16,8))
    plt.suptitle(title, fontsize=20)
    for ind,image_path in enumerate(images_paths):
        image = cv2.imread(image_path)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) 
        try:
            ax.ravel()[ind].imshow(image)
        except:
            continue;
    plt.tight_layout()
    plt.show()
```

```python
def display_groups(df, group_type, lst):
    for item in lst:
        display_multiple_img(path(df,item,group_type)[:9], 3, 3,item)
```

```python
display_groups(train_df,'sp', most_freq_species)
```

```python
display_groups(train_df,'sp', least_freq_species)
```
