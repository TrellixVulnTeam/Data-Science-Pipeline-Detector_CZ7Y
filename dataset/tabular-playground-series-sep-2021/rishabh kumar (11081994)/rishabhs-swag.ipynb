```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='median')
train_imp=imp.fit_transform(train)
```

```python
train=pd.DataFrame(train_imp,columns=train.columns)
train.head()
```

```python
train.isnull().sum()
```

```python
train.shape
```

```python
y=train['claim']
train.describe()
```

```python

train.drop(columns=['id','claim'],axis=1,inplace=True)
```

```python
train.head()
```

```python
import tensorflow as tf
```

```python
print(tf.__version__) 
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
minmax= StandardScaler()
trains=minmax.fit_transform(train)
```

```python
trains.shape
```

```python
 tf.random.set_seed(0)
model = tf.keras.models.Sequential([

  tf.keras.layers.Dense(128, activation='relu',input_shape=(118,)),
  tf.keras.layers.Dense(1,activation='sigmoid')
])
```

```python
model.compile(loss=tf.keras.losses.BinaryCrossentropy(),
             optimizer=tf.keras.optimizers.Adam(),metrics=['accuracy'])
model.fit(trains,y,epochs=20,batch_size=32)
```
