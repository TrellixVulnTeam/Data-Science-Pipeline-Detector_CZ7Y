```python
%matplotlib inline
import cudf, cuml
import cupy as cp

from cuml.manifold import TSNE, UMAP
import matplotlib.pyplot as plt
from matplotlib.pyplot import ylim, xlim
```

```python
train_df = cudf.read_csv('/kaggle/input/tabular-playground-series-sep-2021/train.csv')
train_df.head()
```

```python
train = cp.load('/kaggle/input/tps-09-21-feature-importance-with-xgboost-and-shap/train_oof_shap.npy')
test = cp.load('/kaggle/input/tps-09-21-feature-importance-with-xgboost-and-shap/test_preds_shap.npy')
```

```python
train.shape
```

```python
test.shape
```

```python
train_test = cp.vstack([train, test])
train_test.shape
```

```python
%%time
tsne = TSNE(n_components=2)
train_2D = tsne.fit_transform(train_test)
```

```python
plt.scatter(train_2D[:,0].get(), train_2D[:,1].get(), s = 0.8)
```

```python
%%time
tsne = TSNE(n_components=2)
train_2D = tsne.fit_transform(train)
```

```python
plt.scatter(train_2D[:,0].get(), train_2D[:,1].get(), s = 0.8)
```

```python
plt.scatter(train_2D[:,0].get(), train_2D[:,1].get(), c=train_df['claim'].values.get(), s = 0.8)
```

```python
%%time
umap = UMAP(n_components=2)
train_2D = umap.fit_transform(train)
```

```python
plt.scatter(train_2D[:,0].get(), train_2D[:,1].get(), c=train_df['claim'].values.get(), s = 0.8)
```
