```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
X = pd.read_csv('../input/tabular-playground-series-sep-2021/train.csv', index_col='id')
X_test = pd.read_csv('../input/tabular-playground-series-sep-2021/test.csv', index_col='id')
```

```python
zero = X[X['claim'] == 0].isnull().sum(axis=1)
one = X[X['claim'] == 1].isnull().sum(axis=1)
zero_counts = zero.value_counts()
one_counts = one.value_counts()
prob_one = one_counts / (one_counts + zero_counts)
dict_one = dict(prob_one)
print(dict_one)
```

```python
testnan = X_test.isnull().sum(axis=1).value_counts()
testnan
```

```python
# test has one row with 15 missing values, add to dict
dict_one[15] = 0.5
```

```python
y = X.pop('claim')
```

```python
missing = X_test.isnull().sum(axis=1)
probabilities = np.array([dict_one[x] for x in missing])
```

```python
sample_solution = pd.read_csv('../input/tabular-playground-series-sep-2021/sample_solution.csv')
sample_solution['claim'] = probabilities
sample_solution.to_csv('submission.csv', index=False)
```
