```python
import numpy as np
import pandas as pd
from supervised.automl import AutoML # mljar-supervised
```

```python
train = pd.read_csv("../input/tabular-playground-series-sep-2021/train.csv")
test = pd.read_csv("../input/tabular-playground-series-sep-2021/test.csv")
```

```python
train.head()
```

```python
x_cols = train.columns[1:-1].tolist()
y_col = train.columns[-1]
```

```python
automl = AutoML(
    mode="Compete",
    total_time_limit=7*3600,
    eval_metric="auc"
)
automl.fit(train[x_cols], train[y_col])
```

```python
automl.report()
```

```python
preds = automl.predict_all(test)
```

```python
preds.head()
```

```python
sub = pd.DataFrame({"id": test.id, "claim":preds["prediction_1"]})
```

```python
sub.to_csv("1_submission.csv", index=False)
```
