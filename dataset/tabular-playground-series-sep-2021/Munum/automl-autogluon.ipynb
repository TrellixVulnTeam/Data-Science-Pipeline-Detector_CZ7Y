# Imports

```python
import pandas as pd
import numpy as np
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
train_data = TabularDataset('/kaggle/input/tabular-playground-series-sep-2021/train.csv').drop('id', axis=1)
train_data.head()
```

```python
test_data = TabularDataset('/kaggle/input/tabular-playground-series-sep-2021/test.csv').drop('id', axis=1)
test_data.head()
```

# Train Model

```python
metric = 'roc_auc'
time=7*3600
verbose=1
predictor = TabularPredictor(
    label='claim',
    eval_metric=metric,
    verbosity=verbose,
)
```

```python
%%time
predictor.fit(
    train_data=train_data,
    time_limit=time,
    presets='best_quality',
    verbosity=verbose
)
```

```python
predictor.leaderboard(train_data, silent=True)
```

```python
results = predictor.fit_summary()
```

```python
preds = predictor.predict_proba(test_data)
preds.head(5)
```
