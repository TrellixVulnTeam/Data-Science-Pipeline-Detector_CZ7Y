```python
import numpy as np
import pandas as pd
from sklearn.model_selection import StratifiedKFold, KFold
import matplotlib.pyplot as plt

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
        
input_path = '/kaggle/input/tabular-playground-series-sep-2021/'
```

```python
train = pd.read_csv(input_path + 'train.csv', index_col='id')
print(train.shape)
train.head()
```

```python
train.claim.hist()
```

```python
train['fold'] = -1

kFold = KFold(n_splits=10, shuffle=True, random_state=1234)

for fold, (train_idx, valid_idx) in enumerate(kFold.split(X = train)):
    train.loc[valid_idx, 'fold'] = fold
    
train.head(10)
```

```python
train.fold.value_counts()
```

```python
fig, axes = plt.subplots(nrows=4, ncols=3,figsize=(15,15))

j = 0
for i in range(10):
    train[train.fold == i].claim.hist(ax=axes[j,i%3])
    axes[j, i%3].set_title('Fold {}'.format(i))
    if i%3 == 2:
        j += 1
```

```python
train.to_csv("train_10_folds.csv")
```
