```python
train=train.apply(lambda x: x.fillna(x.mean()), axis=0)
```

```python
y_train=train['claim']
X_train=train.drop(['id','claim'], axis=1)
```

```python
test=pd.read_csv('../input/tabular-playground-series-sep-2021/test.csv')
```
