```python
scaler = StandardScaler()
#train = scaler.fit_transform(train)

#test = scaler.transform(test)
train=pd.DataFrame(scaler.fit_transform(train), columns=all_features)
test = pd.DataFrame(scaler.transform(test), columns=all_features)
print(test)
```

```python
pca = PCA(n_components=110)
pca.fit(train)
X_train_pca = pca.transform(train)
X_test_pca = pca.transform(test)
train = pd.DataFrame(data = X_train_pca)
test = pd.DataFrame(data = X_test_pca)

X=train.copy()
X_test = test.copy()
```

```python
#del test, train, scaler
#gc.collect()
from sklearn.neural_network import MLPClassifier
clf = MLPClassifier(hidden_layer_sizes=10, random_state=1, activation='logistic')
clf.fit(X, y)
Proba1=clf.predict_proba(X)
predictes_for_test=clf.predict_proba(X_test)
print(f'roc_auc_scorettttttttttt: {metrics.roc_auc_score(y, Proba1[:,1])}')
```
