```python
import pandas as pd
import numpy as np

path = '../input/tabular-playground-series-sep-2021/'
sam = pd.read_csv(path+'sample_solution.csv')
train = pd.read_csv(path+'train.csv')
test = pd.read_csv(path+'test.csv')
```

```python
train.describe()
```

```python
#add the 'na_sum' 'std' ''
# fill the Nan 
features_col = list(train.columns[1:119])
target_col = train.columns[-1]

train['na_sum'] = train[features_col].isna().sum(axis = 1)
test['na_sum'] = test[features_col].isna().sum(axis=1)

train[features_col] = train[features_col].fillna(train[features_col].mean())
test[features_col] = test[features_col].fillna(test[features_col].mean())

train['std'] = train[features_col].std(axis=1)
test['std'] = train[features_col].std(axis=1)

features_col += ['na_sum','std']
```

```python
# detect the outliers 
# -1 : outlier , 1 : normal
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import IsolationForest
data_train = train[features_col]
scaler= StandardScaler()
np_scaler = scaler.fit_transform(data_train)
data_train = pd.DataFrame(np_scaler)
# train the IsolationForest 
outlier_fraction = 0.15
model = IsolationForest(contamination=outlier_fraction,random_state=0)
model.fit(data_train)


```

```python
train['outlier'] = pd.Series(model.predict(data_train))
train['outlier'].value_counts()
```

```python
data_test = test[features_col]
scaler = StandardScaler()
np_scaler = scaler.fit_transform(data_test)
data_test = pd.DataFrame(np_scaler)

model.fit(data_test)
test['outlier'] = pd.Series(model.predict(data_test))
test['outlier'].value_counts()
```

```python
train.to_csv('train.csv')
test.to_csv('test.csv')
```
