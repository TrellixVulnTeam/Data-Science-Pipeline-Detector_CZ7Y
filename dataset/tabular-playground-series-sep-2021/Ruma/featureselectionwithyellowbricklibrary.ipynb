```python
target = df1['claim']

features = df1.drop('claim', axis=1)
```

```python
from yellowbrick.target import FeatureCorrelation

feature_names = list(features.columns)
```

The default correlation calculated is the Pearson correlation

```python
from matplotlib.pyplot import figure

figure(figsize=(20,18), dpi=80)
visualizer = FeatureCorrelation(labels = feature_names, method='pearson')

visualizer.fit(features, target)

visualizer.poof()
```

```python
final_features = features[['f48','f47','f46','f45','f42','f41','f39','f38','f36','f35','f34','f32','f31','f24','f21','f14','f13','f1','f2','f3','f5','f6','f7','f8']]
```

```python
final_features.head()
```

```python
final_features.shape
```

```python
from sklearn.model_selection import train_test_split

X = final_features
Y = target

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
x_train.shape, y_train.shape,x_test.shape, y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rnd_clf = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
rnd_clf.fit(x_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,plot_confusion_matrix,accuracy_score
y_pred = rnd_clf.predict(x_test)
print(accuracy_score(y_test, y_pred))
```

```python
rnd_clf.score(x_test, y_test)
```

```python
confusion_matrix(y_test,y_pred)
```
