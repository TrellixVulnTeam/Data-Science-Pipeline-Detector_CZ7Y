```python
import numpy as np
import pandas as pd

import os
from sklearn.model_selection import StratifiedKFold,KFold
```

```python
from pathlib import Path

input_path = Path('../input/tabular-playground-series-sep-2021/')
# Load the training data and intialize constants...
train = pd.read_csv(f"{input_path}/train.csv")
X_test = pd.read_csv(f"{input_path}/test.csv")
X_test.reset_index().to_feather('test_stratfold.ft')
```

```python
TARGET_VAR='claim'
FOLDS=10
```

```python
object_cols = [col for col in train.columns if 'cat' in col]
object_cols_idx = [train.columns.get_loc(col) for col in train.columns if 'cat' in col]
num_cols = list(train.select_dtypes(exclude=['object']).columns)
print(f"total cols {len(train.columns)} objectCols {object_cols} num_cols[Total:{len(num_cols)}] {num_cols}")
# Contains only Numeric cols
```

# Folding to use in other notebooks .

```python
skf = StratifiedKFold(n_splits=FOLDS, shuffle=True, random_state=42)

train['kfold'] = -1

for fold, (train_idx, valid_idx) in enumerate(skf.split(X=train, y=train['claim'])):
    train.loc[valid_idx, 'kfold'] = fold

#train.to_csv('train_stratfold.csv', index=False)
train.head()
```

```python
# write in feather format
train.reset_index().to_feather('train_stratfold.ft')
# write in parquet format 
train.to_parquet('train_stratfold.parquet', index=False)
```
