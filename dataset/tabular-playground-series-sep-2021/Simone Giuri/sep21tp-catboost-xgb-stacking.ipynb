```python
%%time
# Read the data
import datatable as dt  # pip install datatable
sub_cat = dt.fread("../input/sep21tp-catboost-sub-stacking/submission.csv").to_pandas().set_index("id")
sub_xgb = dt.fread("../input/sep21tp-multirnd-stacking-adv/submission.csv").to_pandas().set_index("id")
sub_lgb = dt.fread("../input/single-simple-lightgbm/submission.csv").to_pandas().set_index("id")
```

```python
import statistics
preds =  pd.DataFrame()
preds["sub_cat"] = sub_cat.claim
preds["sub_xgb"] = sub_xgb.claim
preds["sub_lgb"] = sub_lgb.claim
preds
```

```python
predictions = pd.DataFrame()
predictions["id"] = preds.index
predictions["claim"] = list(preds.median(axis=1))

predictions.to_csv('submission.csv', index=False, header=predictions.columns)
predictions.head()
```
