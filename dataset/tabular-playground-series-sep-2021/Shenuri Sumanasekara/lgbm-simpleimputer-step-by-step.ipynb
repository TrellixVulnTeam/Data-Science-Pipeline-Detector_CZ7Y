# Read datasets

```python
train_set = pd.read_csv("../input/tabular-playground-series-sep-2021/train.csv")
test_set = pd.read_csv("../input/tabular-playground-series-sep-2021/test.csv")
train_set.head()
```

# Check null values

```python
train_set.isnull().sum().sort_values(ascending =False)
```

```python
y= train_set['claim']
x= train_set.drop(columns=['claim','id'])
print(len(x.columns))
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x,y, test_size = 0.3, random_state = 60,shuffle=True)
print(len(X_train))
print(len(X_test))
```

# LGBM Regressor

```python

from sklearn.metrics import mean_absolute_error

# Function for comparing different approaches
def score_dataset_LGBM(X_train, X_test, Y_train, Y_test):
    model = LGBMRegressor()
    model.fit(X_train, Y_train)
    preds = model.predict(X_test)
    auc = metrics.roc_auc_score(Y_test, preds)
    print("AUC :",auc)
    return mean_absolute_error(Y_test, preds)
```

# Use SimpleImputer to replace missing values with the mean value of each column.



```python
from sklearn.impute import SimpleImputer

# Imputation
my_imputer = SimpleImputer()
imputed_X_train = pd.DataFrame(my_imputer.fit_transform(X_train))
imputed_X_test = pd.DataFrame(my_imputer.transform(X_test))


imputed_X_train.columns = X_train.columns
imputed_X_test.columns = X_test.columns

print("Mean Absolute Error for LGBMRegressor  :")
print(score_dataset_LGBM(imputed_X_train, imputed_X_test, Y_train, Y_test))
```

# LGBMRegressor

```python
lgbm = LGBMRegressor()
       
lgbm.fit(X_train, Y_train)

test_predict = lgbm.predict(X_test)
```

```python
y_pred_proba = lgbm.predict(X_test)
fpr, tpr, _ = metrics.roc_curve(Y_test,  y_pred_proba)
auc = metrics.roc_auc_score(Y_test, y_pred_proba)
print(auc)
plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```
