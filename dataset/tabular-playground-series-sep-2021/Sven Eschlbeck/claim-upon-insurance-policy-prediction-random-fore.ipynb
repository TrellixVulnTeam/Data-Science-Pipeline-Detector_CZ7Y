# Claim upon Insurance Policy Prediction with Random Forest

#### Loading and Displaying data

```python
import pandas as pd
```

```python
df = pd.read_csv("../input/tabular-playground-series-sep-2021/train.csv")

df.head()
```

#### Pre-Cleaning data

```python
df = df.dropna()
```

#### Defining X and y and Splitting data

```python
from sklearn.model_selection import train_test_split

# Columns used as predictors
X = df.drop(["claim"], axis = 1).values

y = df["claim"].values

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0, test_size = 0.25)
```

#### Applying Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(criterion = "entropy")
model.fit(X_train, y_train)

model.score(X_test, y_test)
```

#### Testing

```python
X_test = pd.read_csv("../input/tabular-playground-series-sep-2021/test.csv")

X_test = X_test.fillna(value = 0)

X_test.head()
```

#### Predicting

```python
predictions = model.predict_proba(X_test)
print(predictions)
```

```python
claim = predictions[:,0]
print(claim)
```

```python
X_test["claim"] = claim
```
