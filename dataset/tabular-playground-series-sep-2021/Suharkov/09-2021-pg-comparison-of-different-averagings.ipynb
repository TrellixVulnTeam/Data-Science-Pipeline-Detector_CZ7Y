# Summarizing

### Take previous submission files as input

```python
laml_no_nans = pd.read_csv('../input/my-output-from-092021-pg/laml_no_nans.csv')
laml_plus_nans = pd.read_csv('../input/my-output-from-092021-pg/laml_plus_nans.csv')
h2o_plus_nans = pd.read_csv('../input/my-output-from-092021-pg/h2o_plus_nans.csv')
last_final = pd.read_csv('../input/my-output-from-092021-pg/final_submission.csv')
```

### Compare predictions

```python
hist_data = [laml_no_nans.claim, laml_plus_nans.claim, h2o_plus_nans.claim, last_final.claim]
group_labels = ['laml without nans', 'laml with nans', 'h2o with nans', 'last final']
fig = ff.create_distplot(hist_data, group_labels, bin_size = 0.3, show_hist = False, show_rug = False)
fig.show()
```

### I do not like this last final at all. Let's try something distinct.

#### First, some exponents > 1 and equal weight of each output

```python
final_sub_2 = last_final.copy()
final_sub_3 = last_final.copy()
final_sub_6 = last_final.copy()
final_sub_9 = last_final.copy()
```

```python
final_sub_2.loc[:, 'claim'] = (laml_no_nans ** 2 + laml_plus_nans ** 2 + h2o_plus_nans ** 2) / 3
final_sub_3.loc[:, 'claim'] = (laml_no_nans ** 3 + laml_plus_nans ** 3 + h2o_plus_nans ** 3) / 3
final_sub_6.loc[:, 'claim'] = (laml_no_nans ** 6 + laml_plus_nans ** 6 + h2o_plus_nans ** 6) / 3
final_sub_9.loc[:, 'claim'] = (laml_no_nans ** 9 + laml_plus_nans ** 9 + h2o_plus_nans ** 9) / 3
```

```python
hist_data = [final_sub_2.claim, final_sub_3.claim, final_sub_6.claim, final_sub_9.claim]
group_labels = ['** 2', '** 3', '** 6', '** 9']
fig = ff.create_distplot(hist_data, group_labels, bin_size = 0.3, show_hist = False, show_rug = False)
fig.show()
```

#### Exponents > 1 and equal weight of each output

```python
final_sub_05 = last_final.copy()
final_sub_033 = last_final.copy()
final_sub_025 = last_final.copy()
final_sub_01 = last_final.copy()
```

```python
final_sub_05.loc[:, 'claim'] = (laml_no_nans ** 1/2 + laml_plus_nans ** 1/2 + h2o_plus_nans ** 1/2) / 3
final_sub_033.loc[:, 'claim'] = (laml_no_nans ** 1/3 + laml_plus_nans ** 1/3 + h2o_plus_nans ** 1/3) / 3
final_sub_025.loc[:, 'claim'] = (laml_no_nans ** 1/4 + laml_plus_nans ** 1/4 + h2o_plus_nans ** 1/4) / 3
final_sub_01.loc[:, 'claim'] = (laml_no_nans ** 1/10 + laml_plus_nans ** 1/10 + h2o_plus_nans ** 1/10) / 3
```

```python
hist_data = [final_sub_05.claim, final_sub_033.claim, final_sub_025.claim, final_sub_01.claim]
group_labels = ['** 1/2', '** 1/3', '** 1/4', '** 1/10']
fig = ff.create_distplot(hist_data, group_labels, bin_size = 0.3, show_hist = False, show_rug = False)
fig.show()
```

### Well, applying both too small or too large exponent makes our PDF move towards zero. Now let's try different combos of weights.

```python
final_sub_w1 = last_final.copy()
final_sub_w2 = last_final.copy()
final_sub_w3 = last_final.copy()
final_sub_w4 = last_final.copy()
```

#### First 4, change H2O and make other two equal.

```python
final_sub_w1.loc[:, 'claim'] = laml_no_nans * 0.45 + laml_plus_nans * 0.45 + h2o_plus_nans * 0.1
final_sub_w2.loc[:, 'claim'] = laml_no_nans * 0.4 + laml_plus_nans * 0.4 + h2o_plus_nans * 0.2
final_sub_w3.loc[:, 'claim'] = laml_no_nans * 0.25 + laml_plus_nans * 0.25 + h2o_plus_nans * 0.5
final_sub_w4.loc[:, 'claim'] = laml_no_nans * 0.1 + laml_plus_nans * 0.1 + h2o_plus_nans * 0.8
```

```python
hist_data = [final_sub_w1.claim, final_sub_w2.claim, final_sub_w3.claim, final_sub_w4.claim]
group_labels = ['version 1', 'version 2', 'version 3', 'version 4']
fig = ff.create_distplot(hist_data, group_labels, bin_size = 0.3, show_hist = False, show_rug = False)
fig.show()
```

#### Change weight of PDF without NaNs

```python
final_sub_w5 = last_final.copy()
final_sub_w6 = last_final.copy()
final_sub_w7 = last_final.copy()
final_sub_w8 = last_final.copy()
```

```python
final_sub_w5.loc[:, 'claim'] = laml_no_nans * 0.1 + laml_plus_nans * 0.45 + h2o_plus_nans * 0.45
final_sub_w6.loc[:, 'claim'] = laml_no_nans * 0.2 + laml_plus_nans * 0.4 + h2o_plus_nans * 0.4
final_sub_w7.loc[:, 'claim'] = laml_no_nans * 0.5 + laml_plus_nans * 0.25 + h2o_plus_nans * 0.25
final_sub_w8.loc[:, 'claim'] = laml_no_nans * 0.8 + laml_plus_nans * 0.1 + h2o_plus_nans * 0.1
```

```python
hist_data = [final_sub_w5.claim, final_sub_w6.claim, final_sub_w7.claim, final_sub_w8.claim]
group_labels = ['version 5', 'version 6', 'version 7', 'version 8']
fig = ff.create_distplot(hist_data, group_labels, bin_size = 0.3, show_hist = False, show_rug = False)
fig.show()
```

### These plots are not so clear without number. Let's estimate perecentage of values > and < 0.5 in each case.

```python
subs = [final_sub_2, final_sub_3, final_sub_6, final_sub_9, 
        final_sub_05, final_sub_033, final_sub_025, final_sub_01, 
        final_sub_w1, final_sub_w2, final_sub_w3, final_sub_w4, 
        final_sub_w5, final_sub_w6, final_sub_w7, final_sub_w8]
version_name = ['squared, equal', 
                'cubic, equal', 
                '^6, equal', 
                '^9, equal', 
                'sqrt, equal', 
                '^1/3, equal', 
                '^1/4, equal', 
                '^1/10, equal', 
                'linear, laml 2*45%, h2o 10%', 
                'linear, laml 2*40%, h2o 20%', 
                'linear, laml 2*25%, h2o 50%', 
                'linear, laml 2*10%, h2o 80%', 
                'linear, NaNs 2*45%, no NaNs 10%', 
                'linear, NaNs 2*40%, no NaNs 20%', 
                'linear, NaNs 2*25%, no NaNs 50%', 
                'linear, NaNs 2*10%, no NaNs 80%']
```

```python
sub_comparison = pd.DataFrame(columns = ['version_name', 'claim_less_than_0.5', 'claim_more_than_0.5'])
for sub in subs:
    sub_comparison.loc[sub_comparison.shape[0]] = [None, len(sub[sub['claim'] < 0.5]) / len(sub), len(sub[sub['claim'] >= 0.5]) / len(sub)]
```

```python
sub_comparison['version_name'] = version_name
```

#### Add train data to this set

```python
train = pd.read_csv('../input/tabular-playground-series-sep-2021/train.csv')
```

```python
sub_comparison.loc[sub_comparison.shape[0]] = ['Train dataset', len(train[train['claim'] < 0.5]) / len(train), len(train[train['claim'] >= 0.5]) / len(train)]
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
ax = sns.barplot(x = 'claim_less_than_0.5', y = 'version_name', data = sub_comparison)
ax.set_title('Amount of claims less than 0.5 with different averaging versions')
plt.axvline(x = 1, color = 'red')
plt.text(1, 10, 'All data', color = 'red', rotation = 90, verticalalignment = 'top', horizontalalignment = 'right')
plt.axvline(x = 1 / 2, color = 'blue')
plt.text(1 / 2, 10, '50% of data', color = 'blue', rotation = 90, verticalalignment = 'top', horizontalalignment = 'right')
plt.xlabel('Percentage')
plt.ylabel('Version of averaging')
plt.show()
```

### I wonder if the model exist which can get straight into that 50% zone, or that will be overfitting?

#### What about average of the first and the last? One on them stands out a little to the right side, the other - to the left. They can balance each other.

```python
balanced = final_sub_2.copy()
balanced.loc[:, 'claim'] = (final_sub_2 + final_sub_w8) / 2
```

```python
sub_comparison.loc[sub_comparison.shape[0]] = ['Balanced try', len(balanced[balanced['claim'] < 0.5]) / len(balanced), len(balanced[balanced['claim'] >= 0.5]) / len(balanced)]
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
ax = sns.barplot(x = 'claim_less_than_0.5', y = 'version_name', data = sub_comparison)
ax.set_title('Amount of claims less than 0.5 with different averaging versions')
plt.axvline(x = 1, color = 'red')
plt.text(1, 10, 'All data', color = 'red', rotation = 90, verticalalignment = 'top', horizontalalignment = 'right')
plt.axvline(x = 1 / 2, color = 'blue')
plt.text(1 / 2, 10, '50% of data', color = 'blue', rotation = 90, verticalalignment = 'top', horizontalalignment = 'right')
plt.xlabel('Percentage')
plt.ylabel('Version of averaging')
plt.show()
```
