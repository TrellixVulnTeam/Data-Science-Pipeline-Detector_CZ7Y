```python

import numpy as np
import pandas as pd 
from sklearn import model_selection
```

```python
df_train = pd.read_csv("../input/tabular-playground-series-sep-2021/train.csv")
```

```python
df_train
```

# Stratified k-fold is used because the dataset is binary classification

```python
#create a new column called kfold and fill it with -1
df_train["kfold"] = -1
```

```python
df_train
```

```python
#randomize the rows of the data
df_train = df_train.sample(frac=1).reset_index(drop=True)
```

```python
#fetch targets(here it is "claim")
y = df_train.claim.values
```

```python
#initiate the kfold class from model_selection module
kf = model_selection.StratifiedKFold(n_splits=5)
```

```python
#fill the new kfold column
for fold, (trn_indices, vld_indices) in enumerate(kf.split(X=df_train, y=y)):
    df_train.loc[vld_indices, 'kfold'] = fold
```
