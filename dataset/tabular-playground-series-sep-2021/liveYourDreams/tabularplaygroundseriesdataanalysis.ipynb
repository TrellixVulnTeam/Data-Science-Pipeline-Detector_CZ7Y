```python
corr=train_data.corr()
corr
```

**We could not display the whole correlation table. Let's drow a plot**

```python
fig, ax = plt.subplots(figsize=(25,25))
sns.heatmap(corr,
            cmap=sns.diverging_palette(230, 10, as_cmap=True),
            square=True,
            ax=ax)
plt.show()
```

**The plot is not clear either. Let's split the train data and cheeck the correlation with 'claim' column**

```python
for i in range(0,11):
    cols=[]
    for j in range (0,10):
        k=i*10+j
        fx='f'+str(k)
        if (fx != 'f0'):
            cols.append(fx)
    print(cols)
    corr1=train_data[cols+['claim']].corr()['claim']
    corr2=train_data[cols+['claim']].corr(method='spearman')['claim']
    print(corr1)
    print(corr2)
    print()
    print()
cols=[]
for i in range(110,119):
    fx='f'+str(i)
    cols.append(fx)
print(cols)
corr1=train_data[cols+['claim']].corr()['claim']
corr2=train_data[cols+['claim']].corr(method='spearman')['claim']
print(corr1)
print(corr2)
```
