```python
import pandas as pd
import numpy as np
from sklearn import model_selection
```

```python
df_train = pd.read_csv('../input/tabular-playground-series-sep-2021/train.csv')
```

```python
df_train.head()
```

```python
df_train.shape
```

```python
df_train.claim.hist()
```

The above graphs shows that its a classification problem

```python
df_train["kfold"] = -1
```

```python
df_train.head()
```

```python
kf = model_selection.KFold(n_splits=5, shuffle= True, random_state = 42)
for fold, (train_indicies, valid_indicies) in enumerate(kf.split(X= df_train)):
    df_train.loc[valid_indicies,"kfold"] = fold
```

```python
df_train.head()
```

```python
df_train.kfold.value_counts()
```

```python
df_train[df_train.kfold == 1].claim.hist()
```

```python
df_train.to_csv("train_folds.csv", index = False)
```
