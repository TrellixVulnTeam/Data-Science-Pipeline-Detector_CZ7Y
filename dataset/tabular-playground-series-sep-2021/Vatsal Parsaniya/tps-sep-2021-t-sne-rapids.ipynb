<div style="background-color:#FFFF00;color:#000000">
    <center><h2>Import</h2></center>
</div>

```python
%matplotlib inline

import numpy as np
import pandas as pd
import seaborn as sns

from matplotlib import pyplot as plt
from cuml.manifold import TSNE
from sklearn.impute import SimpleImputer
from matplotlib.colors import ListedColormap
```

<div style="background-color:#FFFF00;color:#000000">
    <center><h2>Load Data</h2></center>
</div>

```python
train_df = pd.read_csv("../input/tabular-playground-series-sep-2021/train.csv", index_col="id")
test_df = pd.read_csv("../input/tabular-playground-series-sep-2021/test.csv", index_col="id")

display(train_df.head())
display(test_df.head())
```

<div style="background-color:#FFFF00;color:#000000">
    <center><h2>Fill missing value with mean</h2></center>
</div>

```python
features_list = [col for col in train_df.columns if col.startswith('f')]
print("number of features : ", len(features_list))
print("-"*40)
print("missig value before processing in train : ", train_df.isna().sum().sum())
print("missig value before processing in test : ", test_df.isna().sum().sum())
print("-"*40)

imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
train_df[features_list] = imputer.fit_transform(train_df[features_list])
test_df[features_list] = imputer.transform(test_df[features_list])

print("missig value after processing in train : ", train_df.isna().sum().sum())
print("missig value after processing in test : ", test_df.isna().sum().sum())
print("-"*40)
```

<div style="background-color:#FFFF00;color:#000000">
    <center><h2>Apply TSNE on Train Data</h2></center>
</div>

```python
tsne = TSNE(n_components=2, random_state=42)
train_tsne_2c = tsne.fit_transform(train_df[features_list])
```

<div style="background-color:#FFFF00;color:#000000">
    <center><h2>Visualize TSNE on Train Data</h2></center>
</div>

```python
fig, ax1 = plt.subplots(nrows=1,ncols=1,figsize=(15,15))
fig.suptitle("TSNE on Train Data", fontsize=18)
classes = ['0', '1']
colours = ListedColormap(['r','b'])
ax1.set(title="distribution of claim on training data")
scatter = ax1.scatter(train_tsne_2c[:,0], train_tsne_2c[:,1],c=train_df["claim"],cmap=colours,s = 0.7)
ax1.legend(handles=scatter.legend_elements()[0], labels=classes)
plt.show()
```

<div style="background-color:#FFFF00;color:#000000">
    <center><h2>Combine train and test data</h2></center>
</div>

```python
train_df["target"] = 1
test_df["target"] = 0

data_df = pd.concat([train_df[features_list + ["target"]], test_df[features_list + ["target"]]])
display(data_df.tail())
```

<div style="background-color:#FFFF00;color:#000000">
    <center><h2>Apply TSNE on Cmbine Data</h2></center>
</div>

```python
tsne = TSNE(n_components=2, random_state=42)
combine_tsne_2c = tsne.fit_transform(data_df[features_list])
```

<div style="background-color:#FFFF00;color:#000000">
    <center><h2>Visualize TSNE on Train + Test Data</h2></center>
</div>

```python
fig, ax1 = plt.subplots(nrows=1,ncols=1,figsize=(15,15))
fig.suptitle("TSNE on Train + Test Data", fontsize=18)
classes = ['Train', 'Test']
colours = ListedColormap(['r','b'])
scatter = ax1.scatter(combine_tsne_2c[:,0], combine_tsne_2c[:,1],c=data_df["target"],cmap=colours,s = 0.7)
ax1.legend(handles=scatter.legend_elements()[0], labels=classes)
plt.show()
```
