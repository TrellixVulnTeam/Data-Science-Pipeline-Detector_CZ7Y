```python
sub_ag1 = pd.read_csv('../input/tps09-autogluon-081481/TPS09AutoGluon.csv').sort_values('id')
# sub_ag2 = pd.read_csv('../input/d/yus002/tps09-public-notebook/0.81426 ag.csv').sort_values('id') 
sub_lam = pd.read_csv('../input/sep21-lightautoml-starter/In_LightAutoML_we_trust.csv').sort_values('id') 
sub_lgbm1 = pd.read_csv('../input/k/takahiroyoshida012/eda-lightgbm-hyperparameter-tuning-using-optuna/submission.csv').sort_values('id')
sub_lgbm2 = pd.read_csv('../input/tps-sep-2021-single-lgbm/submission.csv').sort_values('id')
sub_xgb = pd.read_csv('../input/xgboost-10fold-cv-blend/submission_blend.csv').sort_values('id')
# sub_nn = pd.read_csv('../input/d/yus002/tps09-public-notebook/0.80493 nn.csv').sort_values('id')
sub_kfold = pd.read_csv('../input/tps-lightgbm-kfold/submission.csv').sort_values('id')
sub_blend = sub_kfold = pd.read_csv('../input/random-weights-blending-tool-tps-sep-2021/rwb 3375 loops version 23.csv').sort_values('id')
```

```python
# Group data together
hist_data = [sub_ag1.claim, sub_lam.claim, sub_lgbm1.claim, sub_lgbm2.claim, sub_xgb.claim, sub_kfold.claim, sub_blend.claim]

group_labels = ['sub_ag1', 'sub_lam', 'sub_lgbm1', 'sub_lgbm2', 'sub_xgb', 'sub_kfold', 'sub_blend']

# Create distplot with custom bin_size
fig = ff.create_distplot(hist_data, group_labels, bin_size=.3, show_hist=False, show_rug=False)
fig.show()
```

```python
data=np.corrcoef([sub_ag1.claim, sub_lam.claim, sub_lgbm1.claim, sub_lgbm2.claim, sub_xgb.claim, sub_kfold.claim, sub_blend.claim])
fig = px.imshow(data,
                x=group_labels,
                y=group_labels
               )
fig.show()
```

# Ensembling

```python
ensemble = sub_ag1.copy()
ensemble.loc[:, 'claim'] = (0 * sub_ag1.claim 
                             + 0 * sub_lam.claim 
                             + 0 * sub_lgbm1.claim 
                             + 0.09 * sub_lgbm2.claim 
                             + 0 * sub_xgb.claim 
                             + 0.21 * sub_kfold.claim
                             + 0.7 * sub_blend.claim)

# Group data together
hist_data = [sub_ag1.claim, sub_lam.claim, sub_lgbm1.claim, sub_lgbm2.claim, sub_xgb.claim, sub_kfold.claim, sub_blend.claim, ensemble.claim]

group_labels = ['sub_ag1', 'sub_lam', 'sub_lgbm1', 'sub_lgbm2', 'sub_xgb', 'sub_kfold', 'sub_blend', 'ensemble']

# Create distplot with custom bin_size
fig = ff.create_distplot(hist_data, group_labels, bin_size=.3, show_hist=False, show_rug=False)
fig.show()
```

```python
ensemble.to_csv('submission.csv', index=False)
```
