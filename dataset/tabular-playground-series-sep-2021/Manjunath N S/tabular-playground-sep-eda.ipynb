Have more that 10% missing values ineach columns,
Le us fill missing values in each column with most repeating mode or median values 

1. Will build model by taking mode i.e taking most repeating values to fill missing place for each feature 
2. 2nd case we can build 2nd model with median values to see how well model behaves

```python
for each_feature in train_master_df.columns:
    if each_feature != "claim" or each_feature != "id":
        train_master_df[each_feature] = train_master_df[each_feature].fillna( stats.mode( train_master_df[each_feature] ).mode[0] ) 
```
