```python
import numpy as np 
import pandas as pd 
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
from sklearn import model_selection
```

```python
df = pd.read_csv("/kaggle/input/tabular-playground-series-sep-2021/train.csv")
```

```python
df["fold"] = -1
```

```python
kf = model_selection.StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
for fold, (train_indicies, valid_indicies) in enumerate(kf.split(X=df,y=df['claim'])):
    df.loc[valid_indicies, "fold"] = fold
```

```python
df.to_csv("train_5folds.csv", index=False)
```
