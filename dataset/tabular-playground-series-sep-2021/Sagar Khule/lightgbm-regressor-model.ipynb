```python
import pandas as pd
train =pd.read_csv("../input/tabular-playground-series-sep-2021/train.csv")
test = pd.read_csv("../input/tabular-playground-series-sep-2021/test.csv")
train = train.drop('id',axis=1)
test = test.drop('id',axis=1)

print("train data : ",len(train))
print("test data : ",len(test))
```

```python
train.isna().sum()
```

```python
test.isna().sum()
```

```python
y_train = train['claim']
X_train = train.drop('claim',axis=1)

from sklearn.impute import SimpleImputer
import numpy as np
imp_median = SimpleImputer(missing_values=np.nan, strategy='median')
X_train = imp_median.fit_transform(X_train)
test = imp_median.transform(test)
```

```python
from lightgbm import LGBMRegressor

model = LGBMRegressor(n_estimators=1000)
model.fit(X_train,y_train)
predict = model.predict(test)
```

```python
sample = pd.read_csv("../input/tabular-playground-series-sep-2021/sample_solution.csv")
sample['claim'] = predict
sample.to_csv("test1.csv",index=None)
sample
```
