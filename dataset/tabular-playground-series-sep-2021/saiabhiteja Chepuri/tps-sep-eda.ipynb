<h2> Checking Missing Values </h2>

```python
missing_values_count = df.isnull().sum()
total_cells = np.product(df.shape)
total_missing = missing_values_count.sum()
percent_missing = (total_missing/total_cells) * 100
print("Missing data in train  % = {}".format(percent_missing))
```

```python
missing_values_count = dftest.isnull().sum()
total_cells = np.product(dftest.shape)
total_missing = missing_values_count.sum()
percent_missing = (total_missing/total_cells) * 100
print("Missing data in test  % = {}".format(percent_missing))
```

```python
#visuvalising the  null values
plt.figure(figsize=(18,9))
sns.heatmap(df.isnull(),yticklabels=False)
plt.title('Heatmap for checking Null Values in train data')
plt.show()
```

```python
#visuvalising the  null values
plt.figure(figsize=(18,9))
sns.heatmap(dftest.isnull(),yticklabels=False)
plt.title('Heatmap for checking Null Values in test data')
plt.show()
```

```python
#getting sum of null values
null_df = df.isnull().sum().sort_values(ascending=False).to_frame()
null_df['percentage'] = (null_df[0]/df.shape[0])*100
null_df=null_df.reset_index()
null_df = null_df.rename(columns = {0:"Missing values","percentage":"Missing_percentage","index":"variable"})
null_df
```

```python
fig = px.bar(data_frame=null_df,y ='variable',x='Missing values',template = 'ggplot2',
             height=5000,text='Missing values',
             title = 'Missing values count of train data')
fig.show()
```

```python
#getting sum of null values
null_df_test = dftest.isnull().sum().sort_values(ascending=False).to_frame()
null_df_test['percentage'] = (null_df_test[0]/df.shape[0])*100
null_df_test=null_df_test.reset_index()
null_df_test = null_df_test.rename(columns = {0:"Missing values","percentage":"Missing_percentage","index":"variable"})
null_df_test
```

```python
fig = px.bar(data_frame=null_df_test,y ='variable',x='Missing values',template = 'ggplot2',
             height=5000,text='Missing values',
             title = 'Missing values count of test data')
fig.show()
```

<h1> Distributions of Data </h1>

```python
sns.set(style='darkgrid')
df1 = df.drop(['id','claim'],axis = 1)
columns = df1.columns
plt.subplots(figsize = (20,80))
plt.suptitle("Distribution of train data")
length = len(columns)
for i,j in zip(columns,range(length)):
    fig = plt.subplot((length/5)+1,5,j+1)
    plt.subplots_adjust(wspace=.10, hspace=.6)
    plt.yticks([])
    plt.grid('off')
    plt.axis('off')
    plt.title("{}".format(i))
    sns.histplot(x=df[i])
```

```python
sns.set(style='darkgrid')
df1 = dftest.drop(['id',],axis = 1)
columns = df1.columns
plt.subplots(figsize = (20,80))
plt.suptitle("Distribution test data")
length = len(columns)
for i,j in zip(columns,range(length)):
    fig = plt.subplot((length/5)+1,5,j+1)
    plt.subplots_adjust(wspace=.10, hspace=.6)
    plt.yticks([])
    plt.grid('off')
    plt.axis('off')
    plt.title("{}".format(i))
    sns.histplot(x=df[i])
```
