```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn import model_selection
```

# Use Feather
[load data with Feather](https://www.kaggle.com/c/tabular-playground-series-sep-2021/discussion/269621)

```python
df_train = pd.read_feather("../input/september-feather/train")
df_train_median = pd.read_feather("../input/september-feather/train_median")
df_train_rg = pd.read_feather("../input/september-feather/train_rg")
```

```python
df_train["kfold"] = -1
df_train_median["kfold"] = -1
df_train_rg["kfold"] = -1
```

```python
kf = model_selection.KFold(n_splits=5, shuffle=True, random_state=92)

def kf_split(df, name):
    for fold,(train_indices, valid_indices) in enumerate(kf.split(X=df)):
        df.loc[valid_indices, "kfold"] = fold
    df.to_feather(name)
```

```python
kf_split(df_train, "train_fold")
kf_split(df_train_median, "train_median_fold")
kf_split(df_train_rg, "train_rg_fold")
```
