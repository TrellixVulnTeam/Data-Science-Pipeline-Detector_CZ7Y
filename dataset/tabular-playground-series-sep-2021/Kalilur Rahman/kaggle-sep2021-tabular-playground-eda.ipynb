```python
from lazypredict.Supervised import LazyClassifier, LazyRegressor

from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split

# Load data and split
X, y = load_boston(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Fit LazyRegressor
reg = LazyRegressor(ignore_warnings=True, random_state=1121218, verbose=False)
models, predictions = reg.fit(X_train, X_test, y_train, y_test) 

models.head(50)
```

### First Description

```python
import seaborn as sns
train.describe().T.style.background_gradient(cmap="nipy_spectral").set_precision(2)
```

```python
train[feature_cols] = train[feature_cols].fillna(train[feature_cols].mean())
test[feature_cols] = test[feature_cols].fillna(test[feature_cols].mean())
```

```python
i = 1
plt.figure()
fig, ax = plt.subplots(30, 4,figsize=(20, 24))
for feature in feature_cols:
    plt.subplot(30, 4,i)
    sns.histplot(train[feature],color="cyan", kde=True,bins=100, label='Train')
    sns.histplot(test[feature],color="pink", kde=True,bins=100, label='Test')
    plt.xlabel(feature, fontsize=9) 
    plt.legend()
    i += 1
```
