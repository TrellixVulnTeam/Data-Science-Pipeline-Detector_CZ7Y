## Import libraries

```python
from pathlib import Path
import numpy as np
import pandas as pd
from sklearn import model_selection
```

## Load the data

```python
data_dir = Path('../input/tabular-playground-series-sep-2021/')
df_train = pd.read_csv(data_dir/'train.csv')
sample_sub = pd.read_csv(data_dir/'sample_solution.csv')
```

## Setup features

```python
target = "claim"
features = [c for c in df_train.columns if c not in ['id', 'claim', 'kfold']]
```

## Create folds

```python
def create_folds(seed=0, splits=5):
    print(f"Creating {splits}-split fold with seed {seed}...")
    df_tmp = df_train.copy()
    df_tmp["kfold"] = -1

    skf = model_selection.StratifiedKFold(n_splits=splits, shuffle=True, random_state=seed)
    for fold, (train_idx, valid_idx) in enumerate(skf.split(X=df_tmp, y=df_tmp[target])):
        df_tmp.loc[valid_idx, "kfold"] = fold

    print("Saving fold...")
    df_tmp.to_csv("train_folds.csv", index=False)
    print("Fold saved.")
```

```python
create_folds(seed=42)
```
