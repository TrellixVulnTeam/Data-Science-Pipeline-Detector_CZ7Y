```python
# Importing core libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import StratifiedKFold
import warnings
warnings.filterwarnings('ignore')

# To impute missing Values
from sklearn.impute import SimpleImputer
```

```python
# Loading data 
df = pd.read_csv("../input/tabular-playground-series-sep-2021/train.csv")

```

```python
imputer = SimpleImputer(strategy='median')
df_train = pd.DataFrame(imputer.fit_transform(df))

# Imputation removed column names hence, getting back the names
df_train.columns = df.columns
```

```python
df_train['kfold'] = -1
```

```python
df_train.columns
```

```python
# Preparing data as a tabular matrix
y_train = df_train.claim
X_train = df_train.drop('claim', axis=1)
```

```python
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=7)
for fold, (train_indicies, valid_indicies) in enumerate(skf.split(X_train,y_train )):
    df_train.loc[valid_indicies, "kfold"] = fold
df_train.to_csv("Tabular_Data_S5Fold.csv", index=False)

```
