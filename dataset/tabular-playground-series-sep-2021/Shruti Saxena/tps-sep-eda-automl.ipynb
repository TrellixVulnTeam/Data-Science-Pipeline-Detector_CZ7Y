##  **Automl**

```python
h2o.init()
```

```python
#import train test files
train = h2o.import_file('../input/tabular-playground-series-sep-2021/train.csv')
test = h2o.import_file('../input/tabular-playground-series-sep-2021/test.csv')
sample_submission = pd.read_csv('/kaggle/input/tabular-playground-series-sep-2021/sample_solution.csv')
```

```python
x = train.columns
y = "claim"
x.remove(y)
x.remove("id") #removing id  
```

```python
aml = H2OAutoML(max_runtime_secs=500, seed=1)
aml.train(x=x, y=y, training_frame=train)
```

```python
lb = aml.leaderboard
lb.head()
```

```python
preds = aml.predict(test)
```

```python
## create submission
sample_submission['claim'] = h2o.as_list(preds)['predict']
sample_submission.to_csv("submission.csv",index=None)
sample_submission.head()
```
