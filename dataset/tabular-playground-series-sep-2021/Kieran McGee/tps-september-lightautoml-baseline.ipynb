This is a basic baseline using LightAutoML preset pipeline. Read the docs [here](https://lightautoml.readthedocs.io/en/latest/)

```python
#Installing LightAutoML
!pip install lightautoml
from lightautoml.automl.presets.tabular_presets import TabularAutoML
from lightautoml.tasks import Task
```

```python
#Reading the data
test = pd.read_csv('../input/tabular-playground-series-sep-2021/test.csv')
train = pd.read_csv('../input/tabular-playground-series-sep-2021/train.csv')
submission = pd.read_csv('../input/tabular-playground-series-sep-2021/sample_solution.csv')
```

```python
train.head()
```

```python
test.head()
```

```python
#Training the model using the preset pipeline.
automl = TabularAutoML(task = Task('binary'))
roles = {'target': 'claim',
        'drop':['id']}
oof_pred = automl.fit_predict(train, roles = roles)
```

```python
#Getting our predictions
test_pred = automl.predict(test)
```

```python
test_pred
```

```python
test_pred_df = pd.DataFrame(test_pred.data)
```

```python
test_pred_df
```

```python
submission['claim'] = test_pred_df
```

```python
submission
```

```python
#Saving our predictions
submission.to_csv('submission', index = False)
```
