```python
import tensorflow as tf
import tensorflow_data_validation as tfdv
print('TFDV version: {}'.format(tfdv.version.__version__))
```

```python
INPUT_DIR = '../input/tabular-playground-series-sep-2021/'

FILE_TRAIN = INPUT_DIR + 'train.csv'
FILE_TEST = INPUT_DIR + 'test.csv'
```

```python
train = tfdv.generate_statistics_from_csv(data_location = FILE_TRAIN)
```

```python
# visualize TRAIN STATS
tfdv.visualize_statistics(train)
```

**Let's compare TRAIN and TEST stats to see if the two Datasets have the same distribution**

```python
test = tfdv.generate_statistics_from_csv(data_location = FILE_TEST)
```

```python
tfdv.visualize_statistics(lhs_statistics = test, 
                          rhs_statistics = train,
                          lhs_name='TEST_DATASET', rhs_name='TRAIN_DATASET')
```

**In the report we find all the differences between the Train and the Test set features' distributions.**

Some conclusion:

* all numeric features
* missing values in each feature hold about 1.6%
* there is no zero value in features
features such as f17, f34, f57, f76, f113, have a gaussian distribution
