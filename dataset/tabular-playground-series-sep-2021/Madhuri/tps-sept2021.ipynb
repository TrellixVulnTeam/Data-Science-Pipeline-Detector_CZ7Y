```python
import numpy as np
import pandas as pd

# loading libraries.
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

# for accuracy
from sklearn.metrics import accuracy_score

# loading dataset
train_data = pd.read_csv("../input/tabular-playground-series-sep-2021/train.csv")
test_data = pd.read_csv("../input/tabular-playground-series-sep-2021/test.csv")

def filling_missing_data(data):
    for col in data.columns:
        if data[col].isnull().sum() > 0:
            data[col].fillna(data[col].median(), inplace=True)
            
filling_missing_data(train_data)
filling_missing_data(test_data)

X = train_data.loc[:, 'f1':'f118']
y = train_data['claim']

def std_scaler(data):
    
    # create an object of StandardScaler
    scale = StandardScaler()
    scale.fit(data)
    scaled_data = scale.transform(data)
    
    return scaled_data
 
# Normalize the train data.
data = std_scaler(X)
Scaled_X = pd.DataFrame(data, columns = X.columns)

# Normalize the test data.
testdata = std_scaler(test_data)
Scaled_test = pd.DataFrame(testdata, columns = test_data.columns)

# spliting data into train set and test set.
X_train, X_test, y_train, y_test =train_test_split(Scaled_X, y, test_size=20, random_state=0)

# define model
def classifier(model):
    
    model.fit(X_train, y_train)
    
    y_hat_test = model.predict(X_test)
    y_hat_train = model.predict(X_train)
    
    testscore = accuracy_score(y_test, y_hat_test)
    trainscore = accuracy_score(y_train, y_hat_train)
    
    print(model)
    print(f"Train Accuracy: {trainscore}")
    print(f"Test Accuracy: {testscore}")

# Logistic Regressor
clf = LogisticRegression()
classifier(clf)
```
