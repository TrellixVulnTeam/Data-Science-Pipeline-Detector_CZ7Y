Creating KFolds as learnt from Abhishek Thakur's videos.

```python
import numpy as np
import pandas as pd
from sklearn import model_selection
```

# Reading the original data

```python
df_train = pd.read_csv("../input/tabular-playground-series-sep-2021/train.csv")
```

```python
df_train.head()
```

```python
df_train.claim.hist()
```

# Creating a new column for kfolds

```python
df_train["kfold"] = -1
```

```python
df_train.head()
```

# Splitting the data into 5 folds

```python
kf = model_selection.KFold(n_splits=5, shuffle=True, random_state=42)
for fold, (train_indicies, valid_indicies) in enumerate(kf.split(X=df_train)):
    df_train.loc[valid_indicies, "kfold"] = fold
```

```python
df_train.head()
```

```python
df_train.kfold.value_counts()
```

```python
df_train.to_csv("train_folds.csv", index=False)
```
