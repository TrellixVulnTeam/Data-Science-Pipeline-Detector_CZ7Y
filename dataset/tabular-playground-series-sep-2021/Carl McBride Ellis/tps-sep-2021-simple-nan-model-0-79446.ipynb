This is a very simple notebook to try out a suggestion by [Lukasz Borecki](https://www.kaggle.com/lukaszborecki) in the topic ["*Add Number of Nans in A Row as a Feature*"](https://www.kaggle.com/c/tabular-playground-series-sep-2021/discussion/270206), giving us a Public Leaderboard score of **0.79446**.

```python
import pandas as pd

test   = pd.read_csv('../input/tabular-playground-series-sep-2021/test.csv')
sample = pd.read_csv('../input/tabular-playground-series-sep-2021/sample_solution.csv')

test['nan_count'] = test.isnull().sum(axis=1)
sample['claim']   = test['nan_count']/test['nan_count'].max()
sample.to_csv('submission.csv',index=False)
```
