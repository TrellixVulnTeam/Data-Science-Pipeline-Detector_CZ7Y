```python
fig, ax = plt.subplots(30,4,figsize=(15,60))
for i in tqdm(range(120)):
    r=i//4
    c=i%4
    sns.histplot(train[train.claim==0][cols[i]], label=cols[i]+' claim=0', ax=ax[r,c], color='black',bins=20)
    sns.histplot(train[train.claim==1][cols[i]], label=cols[i]+' claim=1', ax=ax[r,c], color='C1',bins=20)
    ax[r,c].legend()
    ax[r,c].grid()
plt.show()
```
