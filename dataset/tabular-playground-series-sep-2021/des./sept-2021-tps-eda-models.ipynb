## TPS September 2021

### Introduction

Starting from January this year, the kaggle competition team is offering a month-long tabulary playground competitions. This series aims to bridge between inclass competition and featured competitions with a friendly and approachable datasets.

For this competition, we will predict whether a customer made a claim upon an insurance policy. The ground truth claim is binary valued, but a prediction may be any number from 0.0 to 1.0, representing the probability of a claim. The features in this dataset have been anonymized and may contain missing values.

Files to work with:

- `train.csv` - the training data with the target claim column
- `test.csv` - the test set; you will be predicting the claim for each row in this file
- `sample_submission.csv` - a sample submission file in the correct format

**Evaluation**: 

Submissions are evaluated on area under the ROC curve between the predicted probability and the observed target.

### Modeling:
- I base my modeling on the starter notebook given by the [kaggle competition team](https://www.kaggle.com/ryanholbrook/getting-started-september-2021-tabular-playground). I changed few parameters and imputed NA's and did feature scaling as well.



```python
import pandas as pd
import numpy as np
from pathlib import Path
from xgboost import XGBClassifier
from sklearn.preprocessing import MinMaxScaler
from sklearn.impute import SimpleImputer

from sklearn.model_selection import cross_validate
import warnings 
warnings.filterwarnings('ignore')

data_dir = Path('../input/tabular-playground-series-sep-2021/')

df_train = pd.read_csv(
    data_dir / "train.csv",
    index_col='id',
#     nrows=25000, 
)

X_test = pd.read_csv(data_dir / "test.csv", index_col='id')

FEATURES = df_train.columns[:-1]
TARGET = df_train.columns[-1]

X = df_train.loc[:, FEATURES]
y = df_train.loc[:, TARGET]

seed = 0
fold = 5
```

### Model: xgboost

```python
model_xgb = XGBClassifier(
    max_depth=6,
    eta = 0.4,
    subsample=.85,
    colsample_bytree=.1,
    n_jobs=-1,
    reg_lambda=50,
    reg_alpha=50, 
    gamma=10,
    tree_method='gpu_hist',
    sampling_method='gradient_based', 
    random_state= seed,
)
def score(X, y, model_xgb, cv):
    scoring = ["roc_auc"]
    scores = cross_validate(
        model_xgb, X, y, scoring=scoring, cv=cv, return_train_score=True
    )
    scores = pd.DataFrame(scores).T
    return scores.assign(
        mean = lambda x: x.mean(axis=1),
        std = lambda x: x.std(axis=1),
    )

scores = score(X, y, model_xgb, cv=fold)
display(scores)
```

### Submissions

```python
model_xgb.fit(X, y, eval_metric='auc')

X_test = pd.read_csv(data_dir / "test.csv", index_col='id')

y_pred_xgb = pd.Series(
    model_xgb.predict_proba(X_test)[:, 1],
    index=X_test.index,
    name=TARGET,
)

y_pred_xgb.to_csv("submission_xgb02.csv")
```

### LGBM

```python
from lightgbm import LGBMClassifier
model_lgb = LGBMClassifier(
              num_iterations = 1000,
              objective = "binary", 
              feature_pre_filter = False,
              learning_rate = 0.1,    
              device_type = 'gpu',
    )
def score(X, y, model_lgb, cv):
    scoring = ["roc_auc"]
    scores = cross_validate(
        model_lgb, X, y, scoring=scoring, cv=cv, return_train_score=True
    )
    scores = pd.DataFrame(scores).T
    return scores.assign(
        mean = lambda x: x.mean(axis=1),
        std = lambda x: x.std(axis=1),
    )

scores = score(X, y, model_lgb, cv=fold)
display(scores)
```

#### Submission

```python
model_lgb.fit(X, y, eval_metric='auc')

X_test = pd.read_csv(data_dir / "test.csv", index_col='id')

y_pred_lgb = pd.Series(
    model_lgb.predict_proba(X_test)[:, 1],
    index=X_test.index,
    name=TARGET,
)

y_pred_lgb.to_csv("submission_lgb02.csv")
```

### Catboost

```python
from catboost import CatBoostClassifier
model_catb = CatBoostClassifier(
     iterations = 1000,     
     learning_rate = 0.1,
     random_strength = 20, 
     l2_leaf_reg = 20, 
     depth = 3, 
     subsample = 0.9,
     task_type = 'GPU',
     objective = 'CrossEntropy', 
     bootstrap_type = 'Bernoulli', 
     verbose = 0
)

def score(X, y, model_catb, cv):
    scoring = ["roc_auc"]
    scores = cross_validate(
        model_catb, X, y, scoring=scoring, cv=cv, return_train_score=True
    )
    scores = pd.DataFrame(scores).T
    return scores.assign(
        mean = lambda x: x.mean(axis=1),
        std = lambda x: x.std(axis=1),
    )

scores = score(X, y, model_catb, cv=fold)
display(scores)
```

### submission

```python
model_catb.fit(X, y)

X_test = pd.read_csv(data_dir / "test.csv", index_col='id')

y_pred_catb = pd.Series(
    model_catb.predict_proba(X_test)[:, 1],
    index=X_test.index,
    name=TARGET,
)

y_pred_catb.to_csv("submission_catb02.csv")
```
