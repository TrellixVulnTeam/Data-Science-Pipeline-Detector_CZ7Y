Before we get into more complicated models, let's go with one simple model:
* If there's food adjacent, go to it
* Otherwise, choose randomly from available directions

```python
%%writefile random_hungry.py

from kaggle_environments.envs.hungry_geese.hungry_geese import Observation, Configuration, Action, row_col
import random
dirs = ["NORTH","SOUTH","EAST","WEST"]

def agent(obs,config):
    observation = Observation(obs)
    configuration = Configuration(config)
    player_index = observation.index
    player_goose = observation.geese[player_index]
    player_head = player_goose[0]
    player_row, player_column = row_col(player_head,configuration.columns)
    
    foods = observation.food
    for food in foods:
        
        food_row, food_column = row_col(food, configuration.columns)
        
        if food_row == player_row+1:
            return Action.SOUTH.name
        if food_row + 1 == player_row:
            return Action.NORTH.name
        if food_column == player_column + 1:
            return Action.EAST.name
        if food_column + 1 == player_column:
            return Action.WEST.name
    
    return random.choice(dirs)
```
