```python
import numpy as np
import pickle
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import kaggle_environments
print(kaggle_environments.__file__)
```

```python
obs = {'step': ''}
field_size_a = 7
field_size_b = 11
```

### Paste this code inside agent:

```python
# env.run([agent, 'submission.py', agent , agent]) # W, B, G, R
# my geese in this file dump is second, blue
if obs['step'] == 50: # 51-th turn
    pickle.dump(obs, file=open('obs' + str(obs['step']) + '.dat', 'wb'))
    print('obs' + str(obs['step']))
```

### Load your obs

```python
obs = pickle.load(open("../input/hungry-geese-observation-file-dump-example/obs50.dat", "rb"))
obs
```

### Render field

```python
env = kaggle_environments.make("hungry_geese", debug=False)
env.state[0].observation = obs
env.render(mode="ipython", width=800, height=700)
```

### Reshape field

```python
field = np.ones(field_size_a * field_size_b) * 4 # empty field
for geese in range(len(obs['geese'])):    
    for cell in range(len(obs['geese'][geese])):
        if cell == len(obs['geese'][geese])-1:
            field[obs['geese'][geese][cell]] = geese + 0.2 # indexed geese tail
        elif cell == 0:
            field[obs['geese'][geese][cell]] = geese - 0.2 # indexed geese head
        else:
            field[obs['geese'][geese][cell]] = geese # indexed geese body
```

### Show field

```python
field_reshaped = field.reshape(7, 11)
fig = plt.figure()
heat_map = sns.heatmap(field_reshaped, annot=True, cmap="pink", cbar = False)
fig = heat_map.get_figure()
```

### My head location

```python
i, j = np.where(field_reshaped == obs['index'] - 0.2)
print('my head:', i[0], 'x', j[0], '#', field_reshaped[i[0],j[0]])
```

### Augmented field 3x3

```python
# concatenate 3x7
x = 7
y = 11
augmented_field = np.concatenate([field for _ in range(3)]).reshape(x*3, y)
augmented_field_a = augmented_field.reshape(x*3, y)
```

```python
# show 3x7 augmented field
fig = plt.figure(figsize = (y/2, x*3/2))
heat_map = sns.heatmap(augmented_field_a, annot=True, cmap="pink", cbar = False)
fig = heat_map.get_figure()
```

```python
# concatenate 3x11 field
augmented_field_b = np.concatenate([augmented_field_a for _ in range(3)], axis = 1)
```

```python
# show 3x7, 3x11 augmented field
fig = plt.figure(figsize = (y*3/2, x*3/2))
heat_map = sns.heatmap(augmented_field_b, annot=True, cmap="pink", cbar = False)
fig = heat_map.get_figure()
```

### Augmented head location

```python
i, j = np.where(augmented_field_b == obs['index'] - 0.2)
print('my heads:')
print(i)
print(j)
print('\nmy head centered:')
print(i[4], 'x', j[4], '#', field_reshaped[i[0],j[0]])
```
