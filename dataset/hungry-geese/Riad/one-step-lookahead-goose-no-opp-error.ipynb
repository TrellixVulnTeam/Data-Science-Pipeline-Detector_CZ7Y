```python

def my_agent(obs_dict,config_dict):
    from kaggle_environments.envs.hungry_geese.hungry_geese import Observation,Configuration,Action,row_col
    import numpy as np 
    import random
    global last_action
    def get_heuristic(food_piece,observation,player_index,configuration,Direction='NORTH'):        
        close_opp_segments = 0        
        close_my_goose_segments = 0
        hit_opp_segments = 0
        hit_my_goose_segments = 0
        diff_distance_reward = 0
        if Direction == 'NORTH':

            #player location after one step
            player_goose = observation.geese[player_index]
            player_head = player_goose[0]
            player_row,player_column  = row_col(player_head,configuration.columns)
            next_player_row,next_player_column  = player_row - 1,player_column 
            
            # Food location
            food = food_piece
            food_row,food_column = row_col(food,configuration.columns) 
            # The difference between the two positions
            diff_distance = (food_row - next_player_row)**2 + (food_column - next_player_column)**2
            if diff_distance == 0:
                diff_distance = 1
            diff_distance_reward = 1/diff_distance
            #opp's location
            for i in range(len(observation.geese)): 
                if i == player_index:
                   my_goose = observation.geese[i]
                   my_goose = my_goose[1:]
                   for segment in my_goose:
                      if player_head - 11 == segment:
                          hit_my_goose_segments = hit_my_goose_segments + 1
                        
                   for segment in my_goose:
                      if player_head - 22 == segment:
                          close_my_goose_segments = close_my_goose_segments + 1 
                        
                   for segment in my_goose:
                      if player_head + 11*6  == segment:
                          hit_my_goose_segments = hit_my_goose_segments + 1 
                        
                else:
                    opp_goose = observation.geese[i]
                    for segment in opp_goose:
                       if player_head - 11 == segment:
                          hit_opp_segments = hit_opp_segments + 1 
                            
                    for segment in opp_goose:
                       if player_head + 11*6 == segment:
                          hit_opp_segments = hit_opp_segments + 1 
    
                    for segment in opp_goose:
                       if player_head - 22 == segment:
                          close_opp_segments = close_opp_segments + 1 
                    
                        
            score = 100*diff_distance_reward -50*close_my_goose_segments - 2e2*close_opp_segments - 1e3*hit_opp_segments - 1e3*hit_my_goose_segments 
            print(f'North:{score}')
            return score


        if Direction == 'SOUTH':

            #player location after one step
            player_goose = observation.geese[player_index]
            player_head = player_goose[0]
            player_row,player_column  = row_col(player_head,configuration.columns)
            next_player_row,next_player_column  = player_row + 1,player_column

            # Food location
            food = food_piece
            food_row,food_column = row_col(food,configuration.columns) 
            # The difference between the two positions
            diff_distance = (food_row - next_player_row)**2 + (food_column - next_player_column)**2
            if diff_distance == 0:
                diff_distance = 1
            diff_distance_reward = 1/diff_distance
            #opp segment's location & my goose segments location
            for i in range(len(observation.geese)): 
                if i == player_index:
                   my_goose = observation.geese[i]
                   my_goose = my_goose[1:]
                   for segment in my_goose:
                      if player_head + 11 == segment:
                          hit_my_goose_segments = hit_my_goose_segments + 1
                        
                   for segment in my_goose:
                      if player_head + 22 == segment:
                          close_my_goose_segments = close_my_goose_segments + 1 

                   for segment in my_goose:
                      if player_head - 11*6 == segment:
                          hit_my_goose_segments = hit_my_goose_segments + 1 

                else:
                    opp_goose = observation.geese[i]
                    for segment in opp_goose:
                       if player_head + 11 == segment:
                          hit_opp_segments = hit_opp_segments + 1
                            
                    for segment in opp_goose:
                       if player_head - 11*6 == segment:
                          hit_opp_segments = hit_opp_segments + 1 
                                
                    for segment in opp_goose:
                       if player_head + 22 == segment:
                          close_opp_segments = close_opp_segments + 1                   

            score = 100*diff_distance_reward -50*close_my_goose_segments - 2e2*close_opp_segments - 1e3*hit_opp_segments - 1e3*hit_my_goose_segments 
            print(f'south:{score}')
            return score


        if Direction == 'EAST':

            #player location after one step
            player_goose = observation.geese[player_index]
            player_head = player_goose[0]
            player_row,player_column  = row_col(player_head,configuration.columns)
            next_player_row,next_player_column  = player_row ,player_column + 1

            # Food location
            food = observation.food[0]
            food_row,food_column = row_col(food,configuration.columns) 
            # The difference between the two positions
            diff_distance = (food_row - next_player_row)**2 + (food_column - next_player_column)**2
            if diff_distance == 0:
                diff_distance = 1
            diff_distance_reward = 1/diff_distance        
            #opp segment's location & my goose segments location
            for i in range(len(observation.geese)): 
                if i == player_index:
                   my_goose = observation.geese[i]
                   my_goose = my_goose[1:]
                   for segment in my_goose:
                      if player_head + 1 == segment:
                          hit_my_goose_segments = hit_my_goose_segments + 1 
                        
                   for segment in my_goose:
                      if player_head + 2 == segment:
                          close_my_goose_segments = close_my_goose_segments + 1
                        
                   for segment in my_goose:
                      if player_head - 11 == segment:
                          hit_my_goose_segments = hit_my_goose_segments + 1 

                else:
                    opp_goose = observation.geese[i]
                    for segment in opp_goose:
                       if player_head + 1 == segment:
                          hit_opp_segments = hit_opp_segments + 1
                     
                    for segment in opp_goose:
                       if player_head - 11 == segment:
                          hit_opp_segments = hit_opp_segments + 1 
    
                    for segment in opp_goose:
                       if player_head + 2 == segment:
                          close_opp_segments = close_opp_segments + 1
                        
            score = 100*diff_distance_reward -50*close_my_goose_segments - 2e2*close_opp_segments - 1e3*hit_opp_segments - 1e3*hit_my_goose_segments                             
            print(f'East:{score}')
            return score

        if Direction == 'WEST':

            #player location after one step
            player_goose = observation.geese[player_index]
            player_head = player_goose[0]
            player_row,player_column  = row_col(player_head,configuration.columns)
            next_player_row,next_player_column  = player_row ,player_column - 1

            # Food location
            food = food_piece
            food_row,food_column = row_col(food,configuration.columns) 
            # The difference between the two positions
            diff_distance = (food_row - next_player_row)**2 + (food_column - next_player_column)**2
            if diff_distance == 0:
                diff_distance = 1
            diff_distance_reward = 1/diff_distance
            #opp segment's location & my goose segments location
            for i in range(len(observation.geese)): 
                if i == player_index:
                   my_goose = observation.geese[i]
                   my_goose = my_goose[1:]
                   for segment in my_goose:
                      if player_head - 1 == segment:
                          hit_my_goose_segments = hit_my_goose_segments + 1 
                        
                   for segment in my_goose:
                      if player_head - 2 == segment:
                          close_my_goose_segments = close_my_goose_segments + 1 
                        
                   for segment in my_goose:
                      if player_head + 11 == segment:
                          hit_my_goose_segments = hit_my_goose_segments + 1 

                else:
                    opp_goose = observation.geese[i]
                    for segment in opp_goose:
                       if player_head - 1 == segment:
                          hit_opp_segments = hit_opp_segments + 1
                            
                    for segment in opp_goose:
                       if player_head + 11 == segment:
                          hit_opp_segments = hit_opp_segments + 1 
                                
                    for segment in opp_goose:
                       if player_head - 2 == segment:
                          close_opp_segments = close_opp_segments + 1 
                            
            score = 100*diff_distance_reward -50*close_my_goose_segments - 2e2*close_opp_segments - 1e3*hit_opp_segments - 1e3*hit_my_goose_segments 
            print(f'West:{score}')
            return score
        
        
    def next_move(last_action = None):
            scores = []
            observation = Observation(obs_dict)
            configuration = Configuration(config_dict)
            player_index = observation.index
            player_goose = observation.geese[player_index]
            player_head = player_goose[0]
            player_row,player_column = row_col(player_head,configuration.columns)
            food = observation.food[0]
            food_row,food_column = row_col(food,configuration.columns)
            food_piece = food
            max_score_values = []
            max_score_loc  = []
            
            # Checking allowed movements
            if len(player_goose)== 1:
                if last_action == None:
                        scores.append(get_heuristic(food_piece,observation,player_index,configuration,Direction='SOUTH'))
                        scores.append(get_heuristic(food_piece,observation,player_index,configuration,Direction='NORTH'))            
                        scores.append(get_heuristic(food_piece,observation,player_index,configuration,Direction='EAST'))
                        scores.append(get_heuristic(food_piece,observation,player_index,configuration,Direction='WEST'))            
                        max_score = np.argmax(scores)


                        if max_score == 0:
                            return Action.SOUTH.name 
                        if max_score == 1:
                            return Action.NORTH.name 
                        if max_score == 2:
                            return Action.EAST.name 
                        if max_score == 3:
                            return Action.WEST.name 
                        
                if last_action == 'SOUTH':
                        scores.append(get_heuristic(food_piece,observation,player_index,configuration,Direction='SOUTH'))
                        scores.append(get_heuristic(food_piece,observation,player_index,configuration,Direction='EAST'))
                        scores.append(get_heuristic(food_piece,observation,player_index,configuration,Direction='WEST'))            
                        max_score = np.argmax(scores)


                        if max_score == 0:
                            return Action.SOUTH.name 
                        if max_score == 1:
                            return Action.EAST.name 
                        if max_score == 2:
                            return Action.WEST.name 
                        
                if last_action == 'NORTH':
                        scores.append(get_heuristic(food_piece,observation,player_index,configuration,Direction='NORTH'))            
                        scores.append(get_heuristic(food_piece,observation,player_index,configuration,Direction='EAST'))
                        scores.append(get_heuristic(food_piece,observation,player_index,configuration,Direction='WEST'))            
                        max_score = np.argmax(scores)


                        if max_score == 0:
                            return Action.NORTH.name 
                        if max_score == 1:
                            return Action.EAST.name 
                        if max_score == 2:
                            return Action.WEST.name 

                if last_action == 'EAST':
                        scores.append(get_heuristic(food_piece,observation,player_index,configuration,Direction='SOUTH'))
                        scores.append(get_heuristic(food_piece,observation,player_index,configuration,Direction='NORTH'))            
                        scores.append(get_heuristic(food_piece,observation,player_index,configuration,Direction='EAST'))
                        max_score = np.argmax(scores)

                        if max_score == 0:
                            return Action.SOUTH.name 
                        if max_score == 1:
                            return Action.NORTH.name 
                        if max_score == 2:
                            return Action.EAST.name 
                        
                if last_action == 'WEST':
                        scores.append(get_heuristic(food_piece,observation,player_index,configuration,Direction='SOUTH'))
                        scores.append(get_heuristic(food_piece,observation,player_index,configuration,Direction='NORTH'))            
                        scores.append(get_heuristic(food_piece,observation,player_index,configuration,Direction='WEST'))
                        max_score = np.argmax(scores)

                        if max_score == 0:
                            return Action.SOUTH.name 
                        if max_score == 1:
                            return Action.NORTH.name 
                        if max_score == 2:
                            return Action.WEST.name 
                        
            elif player_head == player_goose[1] + 1 :
                scores.append(get_heuristic(food_piece,observation,player_index,configuration,Direction='SOUTH'))
                scores.append(get_heuristic(food_piece,observation,player_index,configuration,Direction='NORTH'))            
                scores.append(get_heuristic(food_piece,observation,player_index,configuration,Direction='EAST'))  
                max_score = np.argmax(scores)


                if max_score == 0:
                    return Action.SOUTH.name 
                if max_score == 1:
                    return Action.NORTH.name 
                if max_score == 2:
                    return Action.EAST.name 

            elif player_head == player_goose[1] - 1 :
                scores.append(get_heuristic(food_piece,observation,player_index,configuration,Direction='SOUTH'))
                scores.append(get_heuristic(food_piece,observation,player_index,configuration,Direction='NORTH'))            
                scores.append(get_heuristic(food_piece,observation,player_index,configuration,Direction='WEST'))            
                max_score = np.argmax(scores)


                if max_score == 0:
                    return Action.SOUTH.name 
                if max_score == 1:
                    return Action.NORTH.name 
                if max_score == 2:
                    return Action.WEST.name 

            elif player_head == player_goose[1] - 11 :
                scores.append(get_heuristic(food_piece,observation,player_index,configuration,Direction='NORTH'))            
                scores.append(get_heuristic(food_piece,observation,player_index,configuration,Direction='EAST'))            
                scores.append(get_heuristic(food_piece,observation,player_index,configuration,Direction='WEST'))            
                max_score = np.argmax(scores) 

                if max_score == 0:
                    return Action.NORTH.name 
                if max_score == 1:
                    return Action.EAST.name 
                if max_score == 2:
                    return Action.WEST.name 

            elif player_head == player_goose[1] + 11 :
                scores.append(get_heuristic(food_piece,observation,player_index,configuration,Direction='SOUTH'))            
                scores.append(get_heuristic(food_piece,observation,player_index,configuration,Direction='EAST'))            
                scores.append(get_heuristic(food_piece,observation,player_index,configuration,Direction='WEST')) 
                max_score = np.argmax(scores) 

                if max_score == 0:
                    return Action.SOUTH.name 
                if max_score == 1:
                    return Action.EAST.name 
                if max_score == 2:
                    return Action.WEST.name       

            elif player_head == player_goose[1] - 11 * 6 :
                scores.append(get_heuristic(food_piece,observation,player_index,configuration,Direction='SOUTH'))
                scores.append(get_heuristic(food_piece,observation,player_index,configuration,Direction='EAST')) 
                scores.append(get_heuristic(food_piece,observation,player_index,configuration,Direction='WEST'))  
                max_score = np.argmax(scores)


                if max_score == 0:
                    return Action.SOUTH.name 
                if max_score == 1:
                    return Action.EAST.name 
                if max_score == 2:
                    return Action.WEST.name 

            elif player_head == player_goose[1] + 11 * 6 :
                scores.append(get_heuristic(food_piece,observation,player_index,configuration,Direction='NORTH'))
                scores.append(get_heuristic(food_piece,observation,player_index,configuration,Direction='EAST')) 
                scores.append(get_heuristic(food_piece,observation,player_index,configuration,Direction='WEST'))  
                max_score = np.argmax(scores)


                if max_score == 0:
                    return Action.NORTH.name 
                if max_score == 1:
                    return Action.EAST.name 
                if max_score == 2:
                    return Action.WEST.name 

            elif player_head == player_goose[1] - 11 :
                scores.append(get_heuristic(food_piece,observation,player_index,configuration,Direction='SOUTH'))
                scores.append(get_heuristic(food_piece,observation,player_index,configuration,Direction='NORTH'))
                scores.append(get_heuristic(food_piece,observation,player_index,configuration,Direction='EAST')) 
                max_score = np.argmax(scores)


                if max_score == 0:
                    return Action.SOUTH.name 
                if max_score == 1:
                    return Action.NORTH.name 
                if max_score == 2:
                    return Action.EAST.name 

            elif player_head == player_goose[1] + 11 :
                scores.append(get_heuristic(food_piece,observation,player_index,configuration,Direction='NORTH'))
                scores.append(get_heuristic(food_piece,observation,player_index,configuration,Direction='SOUTH')) 
                scores.append(get_heuristic(food_piece,observation,player_index,configuration,Direction='WEST'))  
                max_score = np.argmax(scores)


                if max_score == 0:
                    return Action.NORTH.name 
                if max_score == 1:
                    return Action.SOUTH.name 
                if max_score == 2:
                    return Action.WEST.name 
                
    observation = Observation(obs_dict)
    if observation.step == 0:
        best_move = next_move()
    if observation.step >= 1:
        best_move = next_move(last_action)   
    class last_Direction:
        def __init__(self, direction):
            self.direction = direction
            
    Direction = last_Direction(best_move)
    last_action = Direction.direction
    print(last_action)
    return last_action
```
