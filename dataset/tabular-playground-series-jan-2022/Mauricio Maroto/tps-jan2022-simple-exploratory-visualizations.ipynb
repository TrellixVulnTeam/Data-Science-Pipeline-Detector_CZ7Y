```python
from pathlib import Path
path = Path().cwd().parent / 'input'
files = list(path.rglob('*'))
files
```

```python
train = pd.read_csv(files[2])
train['date'] = pd.to_datetime(train['date'], errors='coerce')
train = train.set_index('date')

test = pd.read_csv(files[3])
test['date'] = pd.to_datetime(test['date'], errors='coerce')
test = test.set_index('date')

train
test
```

```python
# Checking before resample

def len_df(df_wide):
    dfs = {}
    for cat in df_wide['country'].unique():
        for cat_ in df_wide['store'].unique():
            for cat__ in df_wide['product'].unique():
                mask = (df_wide['country']==cat) & (df_wide['store']==cat_) & (df_wide['product']==cat__)
                dfs[cat+cat_+cat__] = df_wide[mask]
                print(df_wide[mask].shape, end=' ')
    return dfs

train_dfs = len_df(train)
```

```python
# All in one

import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(15,12))

for name, df in train_dfs.items():
    _ = ax.plot(df['num_sold'], label=name, linewidth=0.75)
_ = plt.legend()
```

```python
# By Mart

fig, ax = plt.subplots(2, 1, figsize=(15,12))

for name, df in train_dfs.items():
    if 'Mart' in name:
        _ = ax[0].plot(df['num_sold'], label=name, linewidth=0.75)
        _ = ax[0].legend()
    else:
        _ = ax[1].plot(df['num_sold'], label=name, linewidth=0.75)
        _ = ax[1].legend()
```

```python
# By Item

fig, ax = plt.subplots(3, 1, figsize=(15,12))

for name, df in train_dfs.items():
    if 'Sticker' in name:
        _ = ax[0].plot(df['num_sold'], label=name, linewidth=0.75)
        _ = ax[0].legend()
    elif 'Mug' in name:
        _ = ax[1].plot(df['num_sold'], label=name, linewidth=0.75)
        _ = ax[1].legend()
    else:
        _ = ax[2].plot(df['num_sold'], label=name, linewidth=0.75)
        _ = ax[2].legend()
```

```python
# By Country

fig, ax = plt.subplots(3, 1, figsize=(15,12))

for name, df in train_dfs.items():
    if 'Finland' in name:
        _ = ax[0].plot(df['num_sold'], label=name, linewidth=0.75)
        _ = ax[0].legend()
    elif 'Norway' in name:
        _ = ax[1].plot(df['num_sold'], label=name, linewidth=0.75)
        _ = ax[1].legend()
    else:
        _ = ax[2].plot(df['num_sold'], label=name, linewidth=0.75)
        _ = ax[2].legend()
```

```python
# Granular

from itertools import chain, product

fig, ax = plt.subplots(3, 2, figsize=(15,12), sharex=True, sharey=True, constrained_layout=True)
ax = ax.flatten()

for ax_, (c, s) in zip(ax, product(
                    train['country'].unique(),
                    train['store'].unique(),
                    #train['product'].unique()
)):
    for p in train['product'].unique():
        mask = (train['country']==c) & (train['store']==s) & (train['product']==p)
        _ = ax_.plot(train[mask]['num_sold'], label=p, linewidth=0.75)
        _ = ax_.legend()
```
