```python
# Libraries
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
train_data = pd.read_csv('/kaggle/input/tabular-playground-series-jan-2022/train.csv')
test_data = pd.read_csv('/kaggle/input/tabular-playground-series-jan-2022/test.csv')
submission = pd.read_csv('/kaggle/input/tabular-playground-series-jan-2022/sample_submission.csv')

# Rename product because it's a keyword
train_data.rename(columns = {'product': 'product_'},inplace = True)
test_data.rename(columns = {'product': 'product_'},inplace = True)

```

# Data Processing

```python
# Change Date and Create Year Month and Day Features
date_seperated_train= train_data.date.str.rsplit("-", expand=True).rename(columns ={0: 'Year', 1: 'Month', 2: 'Day'})
date_seperated_test= test_data.date.str.rsplit("-", expand=True).rename(columns ={0: 'Year', 1: 'Month', 2: 'Day'})
```

```python
#Create Year, Month, Day columns

train_data['Year'] = date_seperated_train.Year
train_data['Month'] = date_seperated_train.Month
train_data['Day'] = date_seperated_train.Day

test_data['Year'] = date_seperated_test.Year
test_data['Month'] = date_seperated_test.Month
test_data['Day'] = date_seperated_test.Day
```

```python
#Change to datetime
train_data['date'] = pd.to_datetime(train_data['date'], format = "%Y-%m-%d")
test_data['date'] = pd.to_datetime(test_data['date'], format = "%Y-%m-%d")
```

```python
#Create Seasons
train_data['Season']= np.nan

winter_index = train_data[train_data.Month.isin(['01', '02', '12'])].index
spring_index = train_data[train_data.Month.isin(['03', '04', '05'])].index
summer_index = train_data[train_data.Month.isin(['06', '07', '08'])].index
autumn_index = train_data[train_data.Month.isin(['09', '10', '11'])].index

train_data.loc[winter_index, 'Season'] = 'Winter'
train_data.loc[spring_index, 'Season'] = 'Spring'
train_data.loc[summer_index, 'Season'] = 'Summer'
train_data.loc[autumn_index, 'Season'] = 'Autumn'
```
