```python
import os
import numpy as np
import pandas as pd

targetName = 'num_sold'
competitionDir = '/kaggle/input/tabular-playground-series-jan-2022'
submission = pd.read_csv('../input/tabular-playground-series-jan-2022/sample_submission.csv')

preds = []
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        if (dirname != competitionDir) & ('.csv' in filename):
            df = pd.read_csv(os.path.join(dirname, filename))
            if len(df) == len(submission):
                try:
                    preds.append(df[targetName])
                except Exception:
                    pass

submission[targetName] = np.array(preds).mean(axis=0).transpose()
submission.to_csv("submission.csv", index=False)
```
