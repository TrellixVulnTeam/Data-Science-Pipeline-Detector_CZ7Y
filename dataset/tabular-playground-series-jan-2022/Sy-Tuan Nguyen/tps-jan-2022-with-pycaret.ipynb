```python
import sys
sys.path.append('../input/pycaret/pycaret')
sys.path.append('../input/pycaret/pyod')

from pycaret.regression import *

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.simplefilter('ignore')
```

```python
train = pd.read_csv('../input/tabular-playground-series-jan-2022/train.csv')

train.head()
```

```python
date = pd.to_datetime(train.date).dt

train['year']  = date.year
train['month'] = date.month.astype('category')
train['day']   = date.day.astype('category')
train['datofweek'] = date.dayofweek.astype('category')

train.drop(['row_id','date'], axis=1, inplace=True)

train = pd.get_dummies(train)

train.head()
```

```python
model_setup = setup(train, 
                    target = 'num_sold', 
                    session_id=123, 
                    silent=True, 
                    train_size=0.9,
                    fold=10,
                    data_split_shuffle=False,
                    normalize=True,
                    feature_interaction=True
                    ) 
```

```python
def smape(y_pred, y_true):
    return np.mean( np.abs(y_pred - y_true)*200/(y_pred + y_true) )

add_metric('SMAPE', 'SMAPE', smape, greater_is_better = False)

compare_models(sort='SMAPE')
```
