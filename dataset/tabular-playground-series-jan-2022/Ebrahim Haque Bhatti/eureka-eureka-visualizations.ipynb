```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt
import datetime as dt
import warnings
warnings.simplefilter(action="ignore", category=FutureWarning)
```

```python
df=pd.read_csv('/kaggle/input/tabular-playground-series-jan-2022/train.csv')
```

```python
df['date'] = pd.to_datetime(df['date'])
df["Month_Number"] = df["date"].dt.month
df["Week_Day"] = df["date"].dt.dayofweek
df['weekday_name'] = df[['date']].apply(lambda x: dt.datetime.strftime(x['date'], '%A'), axis=1)
df['year'] = df["date"].dt.year
df['day'] = df["date"].dt.day
df['dayofyear'] = df["date"].dt.dayofyear
print(df.info())
```

```python
sns.pairplot(df,hue='product')
plt.show()
```

```python
ff2 = df[(df['date'] >= '2015-01-01') & (df['date'] <= '2018-12-31')]
ff2=ff2.groupby('day').sum()
counts = ff2.groupby('day').sum().reset_index()
sns.lineplot(counts['day'],counts['num_sold'])
plt.show()
```

## **Most sales during start and end of the month**

```python
ff2 = df[(df['date'] >= '2015-01-01') & (df['date'] <= '2018-12-31')]
ff2=ff2.groupby('dayofyear').sum()
counts = ff2.groupby('dayofyear').sum().reset_index()
sns.lineplot(counts['dayofyear'],counts['num_sold'])
plt.show()
```

## **Most sales during start and a big spike at the end of the year**

# **Line Plot by month of products sales in two different stores in three countries**

```python
sns.lineplot(x="Month_Number", y="num_sold", data=df, hue="product" ,palette="Dark2")
plt.show()
```

```python
sns.lineplot(x="Month_Number", y="num_sold", data=df, hue="country" ,palette="Dark2")
plt.show()
```

```python
sns.lineplot(x="Month_Number", y="num_sold", data=df, hue="store" ,palette="Dark2")
plt.show()
```

## **Let's combine the above three plots**

```python
plt.figure(figsize=(21,13))
sns.lineplot(x="Month_Number", y="num_sold", data=df, hue="country",style='product',size=df['store'] ,palette="Dark2")
plt.show()
```
