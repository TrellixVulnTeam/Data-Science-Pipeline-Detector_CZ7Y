```python
train=pd.read_csv('/kaggle/input/tabular-playground-series-jan-2022/train.csv')
test=pd.read_csv('/kaggle/input/tabular-playground-series-jan-2022/test.csv')
```

```python
y=train['num_sold']
train=train.drop(['num_sold'],axis=1)
```

```python
train = train.drop('date', axis=1)
test = test.drop('date', axis=1)

p = {'Kaggle Mug': 0, 'Kaggle Hat': 1, 'Kaggle Sticker': 2}
c = {'Finland': 0, 'Norway': 1,'Sweden': 2}
s = {'KaggleMart': 0, 'KaggleRama': 1}

train['product'] = train['product'].map(p)
train['country'] = train['country'].map(c)
train['store'] = train['store'].map(s)
test['product'] = test['product'].map(p)
test['country'] = test['country'].map(c)
test['store'] = test['store'].map(s)

```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression().fit(train, y)
reg.score(train, y)
```

```python
final = pd.read_csv(
    '/kaggle/input/tabular-playground-series-jan-2022/sample_submission.csv')
final=final.copy()
final['num_sold']=reg.predict(train)[6570]
final.to_csv('submission.csv')
```
