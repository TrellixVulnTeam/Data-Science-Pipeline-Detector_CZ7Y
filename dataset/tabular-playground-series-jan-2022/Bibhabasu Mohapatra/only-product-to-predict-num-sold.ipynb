```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
df = pd.read_csv('../input/tabular-playground-series-jan-2022/train.csv')
df = df.drop(columns=['row_id', 'date', 'country', 'store'])
df_test = pd.read_csv('../input/tabular-playground-series-jan-2022/test.csv')
df_test = df_test.drop(columns=['row_id', 'date', 'country', 'store'])
```

```python
df = pd.get_dummies(df)
df_test = pd.get_dummies(df_test)
```

```python
from sklearn.model_selection import train_test_split
import xgboost
x_train, x_test, y_train, y_test = train_test_split(df[['product_Kaggle Hat', 'product_Kaggle Mug','product_Kaggle Sticker']], df['num_sold'], test_size=0.3)
model = xgboost.XGBRegressor()
model.fit(x_train, y_train)
pred = model.predict(x_test)
def SMAPE(y_true, y_pred):
    denominator = (y_true + np.abs(y_pred)) / 200.0
    diff = np.abs(y_true - y_pred) / denominator
    diff[denominator == 0] = 0.0
    return np.mean(diff)
print(SMAPE(y_test, pred))
```

```python
num_sold = model.predict(df_test)
submission = pd.read_csv('../input/tabular-playground-series-jan-2022/sample_submission.csv')
submission['num_sold'] = num_sold
submission.to_csv('submission.csv', index=False)
```
