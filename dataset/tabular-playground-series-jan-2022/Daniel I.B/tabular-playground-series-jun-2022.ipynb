```python
# import the packages we will need:

import pandas as pd, numpy as np, os, seaborn as sns, datetime as dt
from sklearn.preprocessing import MinMaxScaler 
from sklearn.linear_model import LinearRegression as lg

# get a view of the file directory structure and all available files:

for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
# read the training data file into a Dataframe

df1 = pd.read_csv(r'../input/tabular-playground-series-jan-2022/train.csv')

df1
```

```python
df1.info()
```

```python
# create a helper function to output the feature set given a Dataframe:

def create_feature_set(df):
    # convert object columns to categorical data type:
    categorical_columns = list(df.select_dtypes(include = 'object').columns)
    for column in df[categorical_columns]:
        df[column] = df[column].astype('category')
        
    # convert the date column to ordinal values so we can use them in linear regression:
    df['date'] = df['date'].astype('datetime64')
    df['date'] = df['date'].map(dt.datetime.toordinal)
    
    # replace the values in the categorical columns with their category codes:
    categorical_columns = list(df.select_dtypes(include = 'category').columns)
    for column in categorical_columns:
       df[column] = df[column].cat.codes
    
    # scale the features:
    scaler = MinMaxScaler()
    scaled_data = scaler.fit_transform(df[categorical_columns])
    x = scaled_data
    
    return x
```

```python
# create feature set from the training data and fit the lg model to it:

x = create_feature_set(df1)
y = df1['num_sold']
model = lg()
model.fit(x,y)
```

```python
# read in the test dataset:

df2 = pd.read_csv(r'../input/tabular-playground-series-jan-2022/test.csv')

df2
```

```python
# create a feature set from the test data and predict the values of num_sold

x_test = create_feature_set(df2)

predictions = model.predict(x_test)

answers = pd.DataFrame({'row_id': df2['row_id'], 'num_sold': predictions})

answers
```

```python
# create file to submit

output_file = answers.to_csv('submission.csv', sep=',', index=False)
```
