```python
train.set_index("row_id",inplace=True)
train.head()
```

```python
test.set_index("row_id",inplace=True)
test.head()
```

```python
# change date datatype to datatime 
train['date'] = pd.to_datetime(train.date, format = "%Y-%m-%d")
test['date'] = pd.to_datetime(test.date, format = "%Y-%m-%d")
```

```python
#Add day of week Monday:0 Sunday:6

dayOfWeek={'Monday':0, 'Tuesday':1, 'Wednesday':2, 'Thursday':3, 'Friday':4, 'Saturday':5, 'Sunday':6}
train['day_of_week'] = train['date'].dt.day_name().map(dayOfWeek)
test['day_of_week'] = test['date'].dt.day_name().map(dayOfWeek)
```

```python
train['weekend'] = [1 if a in [4,5,6] else 0 for a in train['day_of_week']]
test['weekend'] = [1 if a in [4,5,6] else 0 for a in test['day_of_week']]
```

```python
train
```

```python
test=test.drop(["date"],axis=1)
test.head()
```

```python
train=train.drop(["date"],axis=1)
train.head()
```

```python
#converting string values(Categorical Values) to integer
train["product"]=train["product"].replace({"Kaggle Mug":0,"Kaggle Hat":1,"Kaggle Sticker":2})
train["store"]=train["store"].replace({"KaggleMart":0,"KaggleRama":1})
train["country"]=train["country"].replace({"Finland":0,"Norway":1,"Sweden":2})
train.head()
```

```python
#converting string values(Categorical Values) to integer in test data
test["product"]=test["product"].replace({"Kaggle Mug":0,"Kaggle Hat":1,"Kaggle Sticker":2})
test["store"]=test["store"].replace({"KaggleMart":0,"KaggleRama":1})
test["country"]=test["country"].replace({"Finland":0,"Norway":1,"Sweden":2})
test.head()
```

# Data Modelling

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(
     train.drop(["num_sold"],axis=1),
     train.num_sold,test_size=0.15,
     random_state=123,shuffle =True)
```

# XG Boost Classifier

```python
# Fitting XGBoost to the training data
from xgboost import XGBClassifier
```

```python

my_model = XGBClassifier(learning_rate=0.3,
                      tree_method='gpu_hist',
                      random_state=0)
my_model.fit(x_train, y_train)

```

```python
# Predicting the Test results
y_pred = my_model.predict(x_test)
```

```python
# validation accuracy
from sklearn.metrics import accuracy_score
print('Accuracy Score : ',accuracy_score(y_test, y_pred))
```

```python
# test prediction
y_pred = my_model.predict(test)
```

```python
#Final submission
submission=pd.DataFrame({"row_id":copy["row_id"],"num_sold":y_pred})
submission.to_csv ('submission.csv',index = False)
```
