# Make Model

Now, Make Model and predict

As you know fbprophet need only datetime and target value

So, we need to classification to use fbprophet

```python
country = train['country'].unique()
store = train['store'].unique()
product = train['product'].unique()

arr = []

for i in range(3):
    for j in range(2):
        for k in range(3):
            arr.append([country[i], store[j], product[k]])
            
m = [Prophet() for i in range(0,18)]
```

I made 18 models. that mean I made all type <country, store, product> 

```python
pred = []

for index, com in enumerate(arr):
    tmp = train[(train['country'] == com[0]) & (train['store'] == com[1]) & (train['product'] == com[2])][['date','num_sold']]
    tmp.columns=['ds','y']
    m[index].fit(tmp)
    future = m[index].make_future_dataframe(periods=365)
    forecast = m[index].predict(future)[['ds', 'yhat']].tail(365)
    pred.append(forecast)    
```

```python
for i in range(18):
    pred[i]['country'] = arr[i][0]
    pred[i]['store'] = arr[i][1]
    pred[i]['product'] = arr[i][2]
    pred[i].columns = ['date',f'yhat_{i}','country','store','product']
    
for i in range(18):
    test = pd.merge(test, pred[i],on = ['date','country','store','product'],how='outer')
    
test = test.fillna(0)
test['target'] = test['yhat_0'] +test['yhat_1'] +test['yhat_2'] +test['yhat_3'] +test['yhat_4'] + \
                test['yhat_5'] +test['yhat_6'] +test['yhat_7'] + test['yhat_8'] +test['yhat_9'] + \
                test['yhat_10'] +test['yhat_11'] +test['yhat_12'] +test['yhat_13'] +test['yhat_14'] + \
                test['yhat_15'] + test['yhat_16']+test['yhat_17']

sample['num_sold'] = test['target']
sample.to_csv("submit.csv", index=None)
```
