This short notebook demonstrates how to add an exchange rate column to the training and test data for TPS Jan 2022 competition.

```python
df_exch_0 = pd.read_csv('../input/eur-nok-sek-exchange-rates-20152019/EUR_NOK_SEK_Exchange_Rates_2015-2019_noNA.csv',parse_dates=['date'])
train_data =  pd.read_csv('../input/tabular-playground-series-jan-2022/train.csv',parse_dates=['date'])
test_data =  pd.read_csv('../input/tabular-playground-series-jan-2022/test.csv',parse_dates=['date'])
```

```python
country_name_dict = {'Euro(EUR)':'Finland','Norwegian krone(NOK)':'Norway','Swedish krona(SEK)':'Sweden'}
df_exch_list = []
for currency in ['Euro(EUR)','Norwegian krone(NOK)','Swedish krona(SEK)']:
    df_exch = df_exch_0[['date',currency]].copy()
    df_exch['country'] = country_name_dict[currency]
    df_exch = df_exch.rename(columns={currency:'exchange_rate'})
    df_exch_list.append(df_exch.set_index(['date','country']))
    
df_exch_1 = pd.concat(df_exch_list)
```

```python
train_data = train_data.join(df_exch_1,on=['date','country'])
test_data = test_data.join(df_exch_1,on=['date','country'])
```
