# Load Time Series Data

```python
# Load Dataset

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
        
filename_training          = '/kaggle/input/tabular-playground-series-jan-2022/train.csv'
```

```python
series = pd.read_csv('/kaggle/input/tabular-playground-series-jan-2022/train.csv', 
                     header = 0,         # We must specify the header information at row 0
                     index_col = 0,      # We hint that the first column contains the index information for the time series
                     parse_dates = True  # We give the function a hint that data in the first column contains dates that need to be parsed
                    )
print(type(series))
```

# Explore Time Series Data

```python
# Peek at the Data
print(series.head())
print(series.tail())

# Number of Observations
print(series.size)

# Querying By Time
# print(series["2018-12-31"]) -- needs to be fixed

# Descriptive Statistics
print(series.describe())
```

- We have 131,490 rows in this Time Series Training Dataset
- We have data for 3 years, i.e. from 01-Jan-2015 to 31-Dec-2018, i.e. 365 days x 3 years = 1,095 days
- We have 18 rows for each day.
- Within each day, we have 6 rows for Finland, 6 rows for Norway and 6 rows for Sweden.
- For each day and each country, we have 3 rows for KaggleMart and 3 rows for KaggleRama.
- For each day and each country and each store type, we have 1 row each for Kaggle Mug, Kaggle Hat and Kaggle Sticker.
