# Import Libraries

```python
from pandas import read_csv
```

# Load Dataset

```python
# Load Dataset

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
        
filename_training          = '/kaggle/input/tabular-playground-series-jan-2022/train.csv'
dataset_training           = read_csv(filename_training)
```

# Print Report

```python
from autoviz.AutoViz_Class import AutoViz_Class

AV = AutoViz_Class()
dftc = AV.AutoViz(
    filename='', 
    sep='' , 
    depVar='num_sold', 
    dfte=dataset_training, 
    header=0, 
    verbose=1, 
    lowess=False, 
    chart_format='png', 
    max_rows_analyzed=4000000, 
    max_cols_analyzed=100
)
```
