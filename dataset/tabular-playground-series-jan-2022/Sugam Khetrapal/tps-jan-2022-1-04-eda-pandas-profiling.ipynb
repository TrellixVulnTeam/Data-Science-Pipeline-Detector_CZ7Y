# Import Libraries

```python
from pandas_profiling import ProfileReport
from pandas import read_csv
```

# Load Dataset

```python
# Load Dataset

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
        
filename_training          = '/kaggle/input/tabular-playground-series-jan-2022/train.csv'
dataset_training           = read_csv(filename_training)
```

# Print Report

```python
report = ProfileReport(dataset_training)
```
