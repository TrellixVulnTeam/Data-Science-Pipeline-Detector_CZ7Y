```python
# Code Sorce : https://www.kaggle.com/bernhardklinger/tps-jan-2022
def pre_process(df):
    
    df['date'] = pd.to_datetime(df['date'])
    df['week']= df['date'].dt.week
    df['year'] = 'Y'+df['date'].dt.year.astype(str)
    df['quarter'] = 'Q'+df['date'].dt.quarter.astype(str)
    df['day'] = df['date'].dt.day
    df['dayofyear'] = df['date'].dt.dayofyear
    df.loc[(df.date.dt.is_leap_year) & (df.dayofyear >= 60),'dayofyear'] -= 1
    df['weekend'] = df['date'].dt.weekday >=5
    df['weekday'] = 'WD' + df['date'].dt.weekday.astype(str)
    df.drop(columns=['date'],inplace=True)   

pre_process(train)
pre_process(test)
```

```python
# Code Sorce : https://www.kaggle.com/bernhardklinger/tps-jan-2022
train.info(), test.info()
```

```python
# Code sorce : https://www.kaggle.com/c/web-traffic-time-series-forecasting/discussion/36414
def SMAPE(y_true, y_pred):
    denominator = (y_true + np.abs(y_pred)) / 200.0
    diff = np.abs(y_true - y_pred) / denominator
    diff[denominator == 0] = 0.0
    return np.mean(diff)
```

```python
# Code Sorce : https://www.kaggle.com/bernhardklinger/tps-jan-2022
reg = setup(data = train,
            target = 'num_sold',
            normalize=True,
            normalize_method='robust', 
            transform_target = True,
            data_split_shuffle = False, 
            create_clusters = False, 
            use_gpu = True,
            silent = True, # silent : If you give a value of 'True', automation is performed, and if you give a value of 'False', no automation is performed.
                                     # ('True'  값을 주면 자동화를 시켜주고,  'False'로 값을 주면 자동화를 시켜주지 않는 역할.)
            fold=10,
            n_jobs = -1)
```

```python
# Code Sorce : https://www.kaggle.com/bernhardklinger/tps-jan-2022
add_metric('SMAPE', 'SMAPE', SMAPE, greater_is_better = False) # TensorFlow is built-in in Pi Carrot, so you can use 'add_metric' of TensorFlow's 'Keras' api.
                                                               # (파이캐롯안에 텐서플로우가 내장 되어있어 텐서플로우의 'Keras' api 의 'add_metric' 사용 가능.)
top =compare_models(sort = 'SMAPE',n_select = 3, include = ['catboost','lightgbm','xgboost']) # Compare Models: Compare different models to find a model that fits your data.
                                                                                              # (Compare Models : 데이터에 적합한 모델을 찾기 위해 다양한 모델을 비교합니다.)
```

```python
# Code Sorce : https://www.kaggle.com/bernhardklinger/tps-jan-2022
blend = blend_models(top) # blend_models( ) : compares several models and finds a suitable model.
                          # (blend_models( )는 여러 모델을 비교해서 적합한 모델을찾아 주는역할을 한다.)
predict_model(blend)
```

```python
# Code Sorce : https://www.kaggle.com/bernhardklinger/tps-jan-2022
final_blend = finalize_model(blend)  # It serves to finalize the model on the entire data set.
                                     # ('finalize'를 번역하면 완성'이라는 의미이기 때문에 전체 데이터 세트에서 모델을 마무리하는 역할을 한다.)
predict_model(final_blend)
```
