# Food for thought

Why do people blend predictions? Scaling is simpler and much more fun! And scaling relentlessly shows if a model is biased.

This time I'll scale (and round) my own [linear model](https://www.kaggle.com/ambrosm/tpsjan22-03-linear-model).

```python
import pandas as pd
sub = pd.read_csv('../input/tpsjan22-03-linear-model/submission_linear_model.csv')
sub.num_sold = (sub.num_sold * 0.9975).round()
sub.to_csv('submission_scaled.csv', index=False)
sub
```
