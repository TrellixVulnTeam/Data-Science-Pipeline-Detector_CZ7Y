```python
import os
import numpy as np
import pandas as pd

targetName = 'num_sold'
competitionDir = '/kaggle/input/tabular-playground-series-jan-2022'
submission = pd.read_csv('../input/tabular-playground-series-jan-2022/sample_submission.csv')
```

```python
preds = []
for dirname, _, filenames in os.walk('/kaggle/input'):
    print(dirname)
    for filename in filenames:
        if (dirname != competitionDir) & ('.csv' in filename):
            df = pd.read_csv(os.path.join(dirname, filename))
            if len(df) == len(submission):
                try:
                    preds.append(df[targetName])
                except Exception:
                    pass
```

```python
random = np.random.RandomState(213)
known = [random.randint(1, 10) for _ in range(10)]
known
```

```python
submission[targetName] = np.average(preds, axis = 0, weights = known)
submission.num_sold = submission.num_sold * .99
submission.to_csv("submission.csv", index=False)
submission
```
