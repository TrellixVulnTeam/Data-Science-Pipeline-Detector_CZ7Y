```python
import pandas as pd
import numpy as np
from supervised import AutoML
```

```python
train = pd.read_csv('../input/tabular-playground-series-jan-2022/train.csv')
test = pd.read_csv('../input/tabular-playground-series-jan-2022/test.csv')
```

```python
# MLJAR AutoML can handle datetime columns automatically :)
train["date"] = pd.to_datetime(train["date"])
test["date"] = pd.to_datetime(test["date"])
```

```python
train
```

```python
x_cols = train.columns[1:-1].tolist()
y_col = "num_sold"
```

```python
x_cols
```

```python
automl = AutoML(mode="Compete", eval_metric="mae")
automl.fit(train[x_cols], train[y_col])
```

```python
p = automl.predict(test)
```

```python
sub = pd.DataFrame({"row_id": test.row_id, "num_sold": p})
sub.to_csv("1_sub.csv", index=False)
```
