```python
import pandas as pd
train=pd.read_csv('../input/tabular-playground-series-jan-2022/train.csv')
train.head()
```

```python
train.info()
```

```python
train.nunique()
```

```python
train.date=pd.to_datetime(train.date, infer_datetime_format=True)
train.info()
```

```python
train.head()
```

```python
from pathlib import Path
```

```python
for i in train.country.unique():
    for j in train.store.unique():
        for k in train['product'].unique():
            data=train[ (train.country== str(i)) &  (train.store==str(j)) & (train['product']==str(k))]
            filename='train'+'_'+i+'_'+j+'_'+k+'.csv'
            p='./'
            data.to_csv(Path(p + filename ), index=False)
```
