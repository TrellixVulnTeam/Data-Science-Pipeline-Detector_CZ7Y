```python
import pandas as pd
train=pd.read_csv('../input/tbs-012022-new-dataset-creation/train_Sweden_KaggleMart_Kaggle Sticker.csv')
train.head()
```

```python
train = train[["date","num_sold"]]
train.head()
```

```python
train= train.rename(columns = {"date":"ds","num_sold":"y"})
train.head()
```

```python
!pip install fbprophet
```

```python
from fbprophet import Prophet
m = Prophet(daily_seasonality = True)
m.fit(train)
```

```python
test=pd.read_csv('../input/tbs-012022-new-dataset-creation/test_Sweden_KaggleMart_Kaggle Sticker.csv')
test.head()
```

```python
future = m.make_future_dataframe(periods=len(test.date.unique()))
prediction = m.predict(future)
prediction.head()
```

```python
import matplotlib.pyplot as plt
m.plot(prediction)
plt.title("Prediction of the Item sold using the Prophet")
plt.xlabel("Date")
plt.ylabel("Num_Sold")
plt.show()
```

```python
from prophet.diagnostics import cross_validation
cutoffs = pd.to_datetime(['2016-01-11','2017-01-11'])
df_cv = cross_validation(m, cutoffs=cutoffs, horizon='365 days')
df_cv.head()
```

```python
from prophet.diagnostics import performance_metrics
df_p = performance_metrics(df_cv)
df_p.head()
```
