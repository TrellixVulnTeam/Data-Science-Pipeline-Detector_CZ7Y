```python
from pycaret.regression import *
```

```python
sol= setup(data =data, session_id=42,
             target = 'num_sold',
                  normalize = True, 
                  transformation = True, 
                  log_experiment = True,
                  remove_multicollinearity = True, 
             transform_target = True,
          combine_rare_levels = True,
            ignore_low_variance = True,
           normalize_method='robust',
            train_size = 0.8,
            handle_unknown_categorical = True, unknown_categorical_method = 'most_frequent',
           date_features = ['date'],
           ignore_features=['row_id'],silent=True
          )
```

```python
model_Voting_Classifier_4 = blend_models(estimator_list = compare_models(n_select = 4))
model_Voting_Classifier_4
```

```python
plot_model(model_Voting_Classifier_4,plot = 'learning')
```

```python
predict_model(model_Voting_Classifier_4)
```

```python
final_Voting_Classifier_4 = finalize_model(model_Voting_Classifier_4)
final_Voting_Classifier_4
```

```python
test=pd.read_csv("../input/tabular-playground-series-jan-2022/test.csv")
test.head()
```

```python
test.date=pd.to_datetime(test.date, infer_datetime_format=True)
test.info()
```

```python
predictions=predict_model(final_Voting_Classifier_4,data =test)
predictions.head()
```

```python
predictions=predictions[['row_id','Label']]
predictions.head()
```
