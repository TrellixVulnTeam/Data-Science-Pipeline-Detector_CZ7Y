```python
train_df = pd.read_csv("/kaggle/input/tabular-playground-series-jan-2022/train.csv")
train_df['date'] = pd.to_datetime(train_df['date'])
train_df.head(3)
```

```python
#Importing important libraries for visualization

import matplotlib.pyplot as plt
import seaborn as sns


#Importing important libraries for data processing

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder,OneHotEncoder, normalize
```

```python
train_df.groupby('country')['num_sold'].sum().sort_values().plot(kind='bar')

plt.title('Total items sold WRT countries')
plt.ylabel('Total items sold')
plt.xlabel('Countries')
plt.show()
plt.tight_layout()
```

```python
plt.figure(figsize=(15,5))
sns.lineplot(train_df['date'],train_df['num_sold'])
plt.show()
```

During the end of each year there is spike in increase in num_sold

```python
unstacked_train_df = train_df.groupby(['country', 'store', 'product']).num_sold.mean().unstack(level='store')
unstacked_train_df['ratio_of_rama_mart'] = unstacked_train_df['KaggleRama']/unstacked_train_df['KaggleMart']
unstacked_train_df
```

KaggleRama is selling 1.7 times than KaggleMart

```python
plt.figure(figsize=(18, 12))
for i, (combi, df) in enumerate(train_df.groupby(['country', 'store', 'product'])):
    ax = plt.subplot(6,3, i+1,ymargin=0.5)
    ax.hist(df['num_sold'],color='pink')
    ax.set_title(combi)
plt.suptitle('Histograms of num_sold', y=1.03)
plt.tight_layout(h_pad=3.0)
plt.show()
```

```python
plt.figure(figsize=(18, 12))
for i, (combi, df) in enumerate(train_df.groupby(['country', 'store', 'product'])):
    ax = plt.subplot(6, 3, i+1, ymargin=0.5)
    sns.lineplot(df['date'],df['num_sold'])
    ax.set_title(combi)
plt.tight_layout(h_pad=3.0)
plt.suptitle('Daily sales for 2015-2018', y=1.03)
plt.show()
```

There is a clear pattern with high sale at end of every year.

```python
#Since we want to check the pattern for Dec month every year, we will filter out that
dec_train_df = train_df[train_df['date'].dt.month == 12].reset_index(drop=True)
dec_train_df.head()
```

```python
#We will check the number of sales in dec every year wrt product
plt.figure(figsize=(10,5))
sns.barplot(x=dec_train_df['date'].dt.year,y=dec_train_df['num_sold'],hue=dec_train_df['product'])
plt.tight_layout()
plt.show()
```

```python
#We will check the number of sales in dec every year wrt country
plt.figure(figsize=(10,5))
sns.barplot(x=dec_train_df['date'].dt.year,y=dec_train_df['num_sold'],hue=dec_train_df['country'])
plt.tight_layout()
plt.show()
```

```python
#We will check the number of sales in dec every year wrt store
plt.figure(figsize=(10,5))
sns.barplot(x=dec_train_df['date'].dt.year,y=dec_train_df['num_sold'],hue=dec_train_df['store'])
plt.tight_layout()
plt.show()
```

```python
#We will check the number of sales in dec for KaggleRama every year wrt country
plt.figure(figsize=(10,5))
sns.barplot(x=dec_train_df[dec_train_df['store']=='KaggleRama']['date'].dt.year,y=dec_train_df\
            [dec_train_df['store']=='KaggleRama']['num_sold'],hue=dec_train_df[dec_train_df['store']=='KaggleRama']['country'])
plt.tight_layout()
plt.show()
```

```python
#We will check the number of sales in dec for KaggleMart every year wrt country
plt.figure(figsize=(10,5))
sns.barplot(x=dec_train_df[dec_train_df['store']=='KaggleMart']['date'].dt.year,y=dec_train_df\
            [dec_train_df['store']=='KaggleMart']['num_sold'],hue=dec_train_df[dec_train_df['store']=='KaggleMart']['country'])
plt.tight_layout()
plt.show()
```

If we try to conclude some of the thins from visualization we can see that -
1. There is spike incerase in sale in December every year - this might be due to the Christmas
2. Norway is having the highest num_sold in comaprison to other two countires
3. KaggleRama is selling more items in comparison to the KaggleMart
4. Kaggle Hat is getting sold most every year
