```python
import pandas as pd
import numpy as np
import os
```

```python
def submission_pathes(path):
    files = os.listdir(path)
    files = [file for file in files if file.split(".")[-1] == "csv"]
    pathes = [os.path.join(path, file) for file in files]
    
    return pathes

def blend(pathes, rounding=True):
    for idx, path in enumerate(pathes):
        submission = pd.read_csv(path).sort_values("row_id")
        predictions = submission["num_sold"].to_numpy()
        
        if idx == 0:
            blended = predictions
            ids = submission["row_id"]
        else:
            blended += predictions
        
    blended /= len(pathes)
        
    if rounding:
        blended = np.round(blended).astype(int)
    
    blended_submission = pd.DataFrame({
        "row_id": ids,
        "num_sold": blended
    })
    
    return blended_submission
```

```python
submission1_pathes = submission_pathes("../input/scaling-is-better-than-blending")
submission2_pathes = submission_pathes("../input/tps-22-eda-prediction")
submission3_pathes = submission_pathes("../input/overfitting-ninja-technique-copied")
submission4_pathes = submission_pathes("../input/tps-jan-2022-eda-pycaret")
submission5_pathes = submission_pathes("../input/tps-jan-2022")
submission6_pathes = submission_pathes("../input/tpsjan22-03-linear-model")
submission7_pathes = submission_pathes("../input/tps-jan-2022-automated-ensembling")
submission8_pathes = submission_pathes("../input/blend-of-blend-of-blend-of-blend-of-blend-of-ble/")

submissions_pathes = submission1_pathes + submission2_pathes + submission3_pathes + submission4_pathes + submission5_pathes + submission6_pathes + submission7_pathes + submission8_pathes
submissions_pathes.remove("../input/overfitting-ninja-technique-copied/wmean4.csv")
```

```python
submission = blend(submissions_pathes, rounding=True)
```

```python
submission.to_csv("submission.csv", index=False)
```
