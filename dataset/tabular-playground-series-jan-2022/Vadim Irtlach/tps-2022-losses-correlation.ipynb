```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import pandas as pd
import numpy as np
import random
import seaborn as sns
```

```python
def SMAPE(y_true, y_pred):
    denominator = (y_true + np.abs(y_pred)) / 200.0
    diff = np.abs(y_true - y_pred) / denominator
    diff[denominator == 0] = 0.0
    return diff.squeeze()
```

```python
def MAE(y_true, y_pred):
    errors = np.abs(y_true - y_pred).squeeze()
    return errors
```

```python
def MSE(y_true, y_pred):
    errors = np.power((y_true - y_pred), 2).squeeze()
    return errors
```

```python
n = 100
for i in range(n):
    random.seed(i)
    np.random.seed(i)
    
    x = np.random.rand(1000000, 1)
    y = np.random.rand(1000000, 1)
    
    mae = MAE(y, x)
    mse = MSE(y, x)
    smape = SMAPE(y, x)
    
    df = pd.DataFrame(np.array([mae, mse, smape]).transpose(1, 0), columns=["MAE", "MSE", "SMAPE"])
    
    p_corr = df.corr("pearson")
    s_corr = df.corr("spearman")
    if i == 0:
        pearson_correlation = p_corr
        spearman_correlation = s_corr
    else:
        pearson_correlation += p_corr
        spearman_correlation += s_corr
        
    if i % 10 == 0:
        print(f"Launch [{i}/{n}]")

pearson_correlation /= n
spearman_correlation /= n
```
