# Simple dataviz

The goal of the notebook is to visualize and get a general understanding of the training.

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
path = "../input/tabular-playground-series-jan-2022/"

train = pd.read_csv(os.path.join(path, 'train.csv'), 
                    index_col = 'row_id',
                    parse_dates = ['date'])
train.head(5)
```

```python
products = list(set(train['product']))
countries = list(set(train.country))
stores = list(set(train.store))

products, countries, stores
```

```python
fig, axs = plt.subplots(3, 3, figsize = (20, 14), sharey = True, sharex = True)

for i, country in enumerate(countries):

    for j, product in enumerate(products):
        ax = axs[i,j]

        for store in stores:

            df = train[(train.store == store) & (train.country == country) & (train["product"] == product)]
            ax.plot(df.date, df.num_sold, label = store)

        ax.set_title(f'{product} sales in {country}')
        
        if (i, j) == (0, 0):
            ax.legend()   
```

# A few observations
1. KaggleRama is obviously selling better than KaggleMart for all products and all countries
2. Hats and mugs are selling better during the summer
3. Some events, most noticeably winter holidays, have a large impact on sales
