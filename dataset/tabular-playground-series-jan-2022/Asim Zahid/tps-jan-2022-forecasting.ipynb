```python

for c in train['country'].unique():
    fig, axes = plt.subplots(3, 3, figsize=(21,15))
    for i, p in enumerate(train['product'].unique()):
        for j, s in enumerate(train['store'].unique()):
            temp = train[(train['country'] == c) & (train['product'] == p) & (train['store'] == s)]
            axes[i, j].plot(temp['num_sold'])
            axes[i, j].title.set_text(f"{c} {p} {s}")
        temp = train[(train['country'] == c) & (train['product'] == p)]
        axes[i, 2].plot(temp['num_sold'])
        axes[i, 2].title.set_text(f"{c} {p} stores")
    plt.plot()
```

```python
temp = train[(train['country'] == 'Finland') & (train['product'] == 'Kaggle Mug') & (train['store'] == 'KaggleMart')]['num_sold']
temp.plot()
temp = temp.rolling(30).mean()
temp.plot()
```

```python
#lets try to remove the general trend
temp = train[(train['country'] == 'Finland') & (train['product'] == 'Kaggle Mug') & (train['store'] == 'KaggleMart')]['num_sold']
temp = temp.to_numpy()
def gradient(p1, p2):
    if (p2[0] - p1[0] == 0):
        print(p1, p2)
    return (p2[1] - p1[1])/(p2[0] - p1[0])
gradients = []
for i in range(365):
    gradients.append(gradient((i, temp[i]), (i + 365*2, temp[i + 365*2])))
gradient_mean = sum(gradients)/len(gradients)
c = 0 #temp.mean()
x = np.linspace(0, 365*4, 365*4 + 1)
print(temp.shape, (gradient_mean*x + c).shape)
plt.plot(temp - temp.mean())
straight_line = gradient_mean*x + c 
plt.plot(x, straight_line)
plt.plot(x, temp - (gradient_mean*x + c))
```

```python
plt.plot(x, fft(temp - straight_line))
```

```python
from scipy.fft import fft, fftfreq, fftshift

import matplotlib.pyplot as plt

t = np.arange(0, 256, 1)
a = np.random.uniform(0, 1, 256)
plt.plot(a + np.sin(0.5*t))
plt.show()
sp = fftshift(fft(a + np.sin(0.5*t)))

freq = fftshift(fftfreq(t.shape[-1]))

plt.plot(freq, sp.real)

plt.show()
```
