### EDA

```python
px.histogram(df_train, x= 'country', color = "store",barmode= 'group', title = "Counts of different kaggle stores across Countries")
```

All 3 countries have equal amount of stores in both the category, so number of store wont be issue since the data is balanced perfectly.

```python
px.histogram(df_train, x= 'country', color = "product", barmode= 'group', title = "Counts of different kaggle Product across Countries")
```

```python
px.histogram(df_train, x= 'store', color = "product", barmode= 'group', title = "Counts of different kaggle Product across stores")
```

We can see that kaggle mug,Hat, Sticker has been sold exactly the same across countries and stories.

```python
px.histogram(x = 'product', y = 'num_sold',data_frame = df_train, title = "Total no of product sold per category")
```

```python
plt.figure(figsize =(12, 10));
sns.displot(df_train, x = "num_sold" ,col= "year",kde= True);

plt.title('Distribution of the number of products sold in each year', y=-0.3, x = - 1.0,fontsize = 20);
```

We can see that roughly in all the year number of sold products is roughly same.

```python
yearly_sales = pd.DataFrame(df_train.groupby(['year'])['num_sold'].sum())
yearly_sales
```

```python
px.bar(yearly_sales, x= yearly_sales.index, y = "num_sold", title = "No of product sold across years")
```

```python
monthly_sales = pd.DataFrame(df_train.groupby(['month'])['num_sold'].sum())
monthly_sales
```

```python
px.bar(monthly_sales, x= monthly_sales.index, y = "num_sold", title = "No of product sold across months")
```

We can see the high sales in the year 2018 and in the month of december.

```python
plt.figure(figsize =(15, 7));
sns.lineplot(x= 'month', y = 'num_sold', data = df_train, hue = 'year',palette = 'flare')
plt.xticks(np.arange(min(df_train['month']), max(df_train['month'])+1, 1.0))
plt.title("Number of sold items across months for different years", fontsize = 20);
plt.show()
```

Here we can see a pattern across all years that, in january it is starting and drops in feb and march , and then increases in april. After april trend seems to decreasing and around october agian the num increases and again in december it increases.

```python
plt.figure(figsize =(15, 7));
sns.lineplot(x= 'month', y = 'num_sold', data = df_train, hue = 'country',palette = 'rocket')
plt.xticks(np.arange(min(df_train['month']), max(df_train['month'])+1, 1.0))
plt.title("Number of sold items across months for different countries", fontsize = 20);
plt.show()
```
