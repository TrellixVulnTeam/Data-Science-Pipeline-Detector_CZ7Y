```python
df['month'] = df['date'].dt.month

month_names = list(calendar.month_name)[1:]

month_store_sell_df = df.groupby(['store', 'month'])['num_sold'].sum().unstack().T.reset_index()
month_store_sell_df.month = month_names
```

```python
heatmap2_data = pd.pivot_table(df, values='num_sold', index=['store'], columns='month')
heatmap2_data.columns = month_names

plt.subplots(figsize = (12,4), dpi = 80)
sns.heatmap(heatmap2_data, cmap="YlGnBu");
```

```python
fig, ax = plt.subplots(figsize = (12.5,5), dpi = 80)

ax=sns.barplot(y=month_store_sell_df['KaggleMart'], x=month_store_sell_df['month'], palette='rainbow')
ax1 = ax.twinx()
sns.lineplot(data=month_store_sell_df, marker='o', x='month', y='KaggleMart', color = 'white', alpha = .8)

total = sum(month_store_sell_df['KaggleMart'])
for i in ax.patches:
    ax.text(i.get_x()+0.06, i.get_height()-20000, \
            str(round((i.get_height()*100)/total, 2))+'%', fontsize=11,weight='bold',color='white')

plt.title('\nSell in KaggleMart over the Year (2015-2018)\n', size=20, color='grey')
ax.axes.set_ylim(-10000, 400000)
ax1.axes.set_ylim(-10000, 500000)

ax1.axes.yaxis.set_visible(False)
ax.set_xlabel('\nMonths\n', fontsize=15, color='grey')
ax.set_ylabel('\nNumber of Sell\n', fontsize=15, color='grey')

for i in ['top', 'right']:
    ax.spines[i].set_visible(False)
    ax1.spines[i].set_visible(False)

for i in  [ax, ax1]:
    i.spines['bottom'].set_bounds(0.005, 11)
    i.spines['left'].set_bounds(0, 400000)
```

```python
fig, ax = plt.subplots(figsize = (12.5,5), dpi = 80)

cmap = matplotlib.cm.get_cmap('plasma', 7)   
clrs = [matplotlib.colors.rgb2hex(cmap(i)) for i in range(cmap.N)]

ax=sns.barplot(y=day_store_sell_df['KaggleMart'], x=day_store_sell_df['day'], palette=clrs)
ax1 = ax.twinx()
sns.lineplot(data=day_store_sell_df, marker='o', x='day', y='KaggleMart', color = 'white', alpha = .8)

total = sum(day_store_sell_df['KaggleMart'])
for i in ax.patches:
    ax.text(i.get_x()+0.1, i.get_height()-50000, \
            str(round((i.get_height()*100)/total, 2))+'%', fontsize=15,weight='bold',
                color='white')

plt.title('\nSell in KaggleMart store over the week (2015-2018)\n', size=20, color='grey')
ax.axes.set_ylim(-10000, 700000)
ax1.axes.set_ylim(-10000, 900000)

ax1.axes.yaxis.set_visible(False)
ax.set_xlabel('\nDay of the Week\n', fontsize=15, color='grey')
ax.set_ylabel('\nNumber of Sell\n', fontsize=15, color='grey')

for i in ['top', 'right']:
    ax.spines[i].set_visible(False)
    ax1.spines[i].set_visible(False)

for i in  [ax, ax1]:
    i.spines['bottom'].set_bounds(0.005, 6)
    i.spines['left'].set_bounds(0, 700000)
```
