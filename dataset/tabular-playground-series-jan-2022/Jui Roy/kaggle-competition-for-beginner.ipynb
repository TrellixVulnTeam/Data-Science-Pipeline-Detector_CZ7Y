**DATA VISUALIZATION**

```python
df_train['date'] = pd.to_datetime(df_train['date'])
df_test['date'] = pd.to_datetime(df_test['date'])
```

```python
df_train.info
```

```python
#Monthly trend
df_train.monthly = df_train.set_index('date').groupby([pd.Grouper(freq = 'M')])[['num_sold']].mean()

plt.figure(figsize = (12, 7))
sns.lineplot(x = 'date', y = 'num_sold', data = df_train, label = 'daily')
sns.lineplot(x = 'date', y = 'num_sold', data = df_train.monthly, label = 'monthly mean', color = 'red')
plt.title('Monthly Trend')
plt.grid(alpha = 0.5)
plt.show()
```

```python
# day of week
df_train['dayofweek'] = df_train['date'].dt.dayofweek
df_test['dayofweek'] = df_test['date'].dt.dayofweek
```

```python
df_train['weekend'] = df_train['dayofweek'].apply(lambda x : x >= 5)
df_test['weekend'] = df_train['dayofweek'].apply(lambda x : x >= 5)
df_train.weekend = df_train.set_index('date').groupby([pd.Grouper(freq = 'M'), 'weekend'])[['num_sold']].mean()

plt.figure(figsize = (10, 7))
sns.lineplot(x = 'date', y = 'num_sold', hue = 'weekend', data = df_train.weekend)
plt.title('Weekend vs. Weekday Trend Comparison')
plt.grid(alpha = 0.5)
plt.show()
```
