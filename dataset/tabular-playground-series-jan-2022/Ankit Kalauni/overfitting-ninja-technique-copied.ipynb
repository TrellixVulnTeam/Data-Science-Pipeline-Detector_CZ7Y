```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

![](https://pbs.twimg.com/media/FIJqLhUVkAIaxNx?format=jpg&name=900x900)

```python
targetName = 'num_sold'
competitionDir = '/kaggle/input/tabular-playground-series-jan-2022'
submission = pd.read_csv('../input/tabular-playground-series-jan-2022/sample_submission.csv')
```

# Import any number of public notebooks to automatically update the ensemble prediction

* Forked from https://www.kaggle.com/sytuannguyen/tps-jan-2022-automated-ensembling/notebook

```python
preds = []
for dirname, _, filenames in os.walk('/kaggle/input'):
    for count, filename in enumerate(filenames):
        if (dirname != competitionDir) & ('.csv' in filename):
            print(f"File: {filename} of {dirname.split('/')[-1]}")
            df = pd.read_csv(os.path.join(dirname, filename))
            if len(df) == len(submission):
                try:
                    preds.append(df[targetName])
                except Exception:
                    pass
```

# Save ensemble prediction to csv

```python
# np.array(preds)[1] # 1 - 6.0 public lb (blend file)
# np.array(preds)[1] # 1 - 5.00896 public lb
# np.array(preds)[2] # 2 - 4.93924 public lb
# np.array(preds)[3] # 3 - 4.92669 public lb

w = [5, 15, 30, 50]
wmean = ((np.array(preds)[0] * w[0]) + (np.array(preds)[1] * w[1]) + (np.array(preds)[2] * w[2]) + (np.array(preds)[3] * w[3])) / sum(w) #weighted avg

wmean1 = ((np.array(preds)[1]) + (np.array(preds)[2]) + (np.array(preds)[3])) / 3 #simple mean

wmean2 = ((np.array(preds)[0]) + (np.array(preds)[1]) + (np.array(preds)[2]) + (np.array(preds)[3])) / 4 #simple mean

w = [10, 20, 30, 40]
wmean3 = ((np.array(preds)[0] * w[0]) + (np.array(preds)[1] * w[1]) + (np.array(preds)[2] * w[2]) + (np.array(preds)[3] * w[3])) / sum(w)

w = [10, 20, 30, 40]
wmean4 = (((np.array(preds)[0] * w[0]) ** 2) + ((np.array(preds)[1] * w[1]) ** 2) + ((np.array(preds)[2] * w[2]) ** 2) + ((np.array(preds)[3] * w[3]) ** 2)) / sum(w)

w = [10, 20, 30, 40]
wmean5 = ((np.array(preds)[0] * w[0]) + (np.array(preds)[1] * w[1]) + (np.array(preds)[2] * w[2]) + (np.array(preds)[3] * w[3]) )/ sum(w)

w = [0, 0, 40, 60]
wmean6 = ((np.array(preds)[0] * w[0]) + (np.array(preds)[1] * w[1]) + (np.array(preds)[2] * w[2]) + (np.array(preds)[3] * w[3]) )/ sum(w)

#formula for weighted-mean = (pred1 * weight1) + (pred2 * weight2) + (pred3 * weight3) + (predn * weightn) / (sumof weights)
```

```python
submission[targetName] = wmean
submission.to_csv("wmean.csv", index=False)

submission[targetName] = wmean1
submission.to_csv("wmean1.csv", index=False)

submission[targetName] = wmean2
submission.to_csv("wmean2.csv", index=False)

submission[targetName] = wmean3
submission.to_csv("wmean3.csv", index=False)

submission[targetName] = wmean4
submission.to_csv("wmean4.csv", index=False)

submission[targetName] = wmean5
submission.to_csv("wmean5.csv", index=False)

submission[targetName] = wmean6
submission.to_csv("wmean6.csv", index=False)

submission[targetName] = wmean2 * 0.98
submission.to_csv("scale1.csv", index=False)

submission[targetName] = np.array(preds)[3] * 0.98
submission.to_csv("scale2.csv", index=False)

submission[targetName] = np.array(preds)[3] * 1.02
submission.to_csv("scale3.csv", index=False)
```

```python
fig = plt.figure(figsize=(18,8), dpi=600)
plot = sns.kdeplot(wmean2, color ='red', label='blend')
sns.kdeplot(wmean2 * 0.98, color ='green',ax = plot, label='blend scaled')
sns.kdeplot(np.array(preds)[3], color ='blue',ax = plot, label='best lb')
sns.kdeplot(np.array(preds)[3] * 0.98, color ='pink',ax = plot, label='best lb scaled')
sns.kdeplot(np.array(preds)[3] * 1.02, color ='yellow',ax = plot, label='best lb scaled')
plt.legend()
plt.xlim([0, 1500])

plt.show()
```
