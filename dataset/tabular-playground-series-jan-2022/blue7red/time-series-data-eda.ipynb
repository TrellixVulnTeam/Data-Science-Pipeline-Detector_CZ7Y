```python
import numpy as np
import pandas as pd 
```

```python
TARIN_PATH = "../input/tabular-playground-series-jan-2022/train.csv"

TARGET = "num_sold"

# time series data common new feature  
DATE = "date"

YEAR = "year"
MONTH = "month"
DAY = "day"

DAYOFYEAR = "dayofyear"
DAYOFMONTH = "dayofMonth"
DAYOFWEEK = "dayofweek"
WEEKDAY = "weekday"
```

```python
train = pd.read_csv(TARIN_PATH)
train.shape
```

# convert object to datetime

```python
train[DATE] = pd.to_datetime(train[DATE])
```

# feature engineering

```python
train[YEAR] = train[DATE].dt.year
train[MONTH] = train[DATE].dt.month
train[DAY] = train[DATE].dt.day

train[DAYOFYEAR] = train[DATE].dt.dayofyear
train[DAYOFMONTH] = train[DATE].dt.days_in_month
train[DAYOFWEEK] = train[DATE].dt.dayofweek
train[WEEKDAY] = train[DATE].dt.weekday
```

```python
train.head()
```

# describe numeric column data 

```python
train.describe()
```

# decribe object column data

```python
trainObject = train.describe(include="object") # same expression => train.describe(include="O")
trainObject
```

```python
trainObject.columns.tolist()
```

# get unique data 

```python
for col in trainObject.columns.tolist():
    print(col," : ",train[col].unique())
```

# get unique data count

```python
for col in trainObject.columns.tolist():
    print("[" + col + "]")
    print(train[col].value_counts())
    print("")
```

# pivot data 

#### index : DATE
