# Importing the dataset

```python
train = pd.read_csv('/kaggle/input/cat-in-the-dat/train.csv')
test = pd.read_csv("/kaggle/input/cat-in-the-dat/test.csv")
print("train and test dataset imported")
```

```python
train.info()
```

```python
#X = train.drop(['id','target'], axis = 1)

X = train[['bin_0','bin_1','bin_2','bin_3','bin_4','nom_0','nom_1','nom_2','nom_3','nom_4']]
y = train['target']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.30,  stratify = y)
```

```python
X_train = pd.get_dummies(X_train, drop_first = True)
```

```python
X_val = pd.get_dummies(X_val, drop_first = True)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression(random_state=0)
clf.fit(X_train,y_train)
```

```python
print(f'Train : {clf.score(X_train, y_train):.3f}')
print(f'Validation : {clf.score(X_val, y_val):.3f}')
```

```python
X_test = test[['bin_0','bin_1','bin_2','bin_3','bin_4','nom_0','nom_1','nom_2','nom_3','nom_4']]
```

```python
X_test = pd.get_dummies(X_test)
```

```python
test_prediciton = clf.predict(X_test[X_train.columns])
```

```python
final_sub = pd.DataFrame({'id': test['id'], 'target':test_prediciton })
final_sub.head()
```
