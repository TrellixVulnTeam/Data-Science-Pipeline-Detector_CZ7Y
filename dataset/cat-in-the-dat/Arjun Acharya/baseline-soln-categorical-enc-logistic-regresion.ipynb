# Import Data

```python
train = pd.read_csv('/kaggle/input/cat-in-the-dat/train.csv')
test = pd.read_csv('/kaggle/input/cat-in-the-dat/test.csv')
```

```python
train.info()
```

```python
test.info()
```

## create train test split

```python
X = train[['bin_0','bin_1','bin_2','bin_3','bin_4','nom_0','nom_1','nom_2','nom_3','nom_4']]
y = train['target']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.30, random_state=101, stratify = y)
```

## One Hot Encoding

```python
X_train = pd.get_dummies(X_train, drop_first = True)
```

```python
X_val = pd.get_dummies(X_val, drop_first = True)
```

```python
X_train.head()
```

## Build Model

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression(random_state=0)
```

```python
clf.fit(X_train,y_train)
```

```python
print(f'Train : {clf.score(X_train, y_train):.3f}')
print(f'Validation : {clf.score(X_val, y_val):.3f}')
```

## Predict Test

```python
X_test = test[['bin_0','bin_1','bin_2','bin_3','bin_4','nom_0','nom_1','nom_2','nom_3','nom_4']]
```

```python
X_test = pd.get_dummies(X_test)
```

```python
X_train.columns
```

```python
test_prediction = clf.predict(X_test[X_train.columns])
```

## Create Submission File

```python
final_sub = pd.DataFrame({'id': test['id'], 'target':test_prediction })
final_sub.head()
```

```python
final_sub.to_csv("Base_Solution.csv", index = False)
```
