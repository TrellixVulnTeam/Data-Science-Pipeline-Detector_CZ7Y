```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
```

```python
df_train = pd.read_csv('../input/cat-in-the-dat/train.csv')
```

```python
df_train.head()
```

```python
features = df_train.drop(['target'], axis = 1)
target = df_train['target']
```

```python
print(features.shape,target.shape)
```

Before getting into encoding,I will just breif you with types data variables present in this data:

1. Binary data : A binary variable a variable that has only 2 values..ie 0/1
2. Categorical data : A categorical variable is a variable that can take some limited number of values.for    example,day of the week.It can be one of 1,2,3,4,5,6,7 only.
3. Ordinal data : An ordinal variable is a categorical variable that has some order associated with it.for example,the ratings that are given to a movie by a user.
4. Nominal data : Nominal value is a variable that has no numerical importance,such as occupation,person name etc..
5. Timeseries data : Time series data has a temporal value attached to it, so this would be something like a date or a time stamp that you can look for trends in time.

```python
df_feature = pd.DataFrame()
le = LabelEncoder()
for c in features.columns:
    df_feature[c] = le.fit_transform(features[c])
    
```

```python
#df_feature.head()
print(df_feature.shape)
```

```python
train_f, test_f, train_y, test_y = train_test_split(df_feature,target,test_size=0.2, random_state=42)
```

```python
print(train_f.shape, train_y.shape)
```

```python
clf = RandomForestClassifier(n_estimators = 200, n_jobs = -1, verbose = 2)
clf.fit(train_f,train_y)
target_pre = clf.predict(test_f)
```

```python
print(accuracy_score(target_pre,test_y))
```
