```python
df_train = df_train.dropna()
```

```python
for column in df_train:
    print(column, " unique value length: ", len(df_train[column].unique()))
```

```python
df_train['bin_0'].value_counts().plot(kind='bar')
```

```python
df_train = df_train.drop(['bin_0'], axis='columns')
```

```python
df_train['bin_1'].value_counts().plot(kind='bar')
```

```python
df_train = df_train.drop(['bin_1'], axis='columns')
```

```python
df_train['bin_3'].replace({"T" : 1, "F" : 0}, inplace=True)
df_train['bin_4'].replace({"Y" : 1, "N" : 0}, inplace=True)
```

```python
merged = df_train.copy()
for i in range(0,8):
  column = f'nom_{i}'
  dummies_temp = pd.get_dummies(merged[column])
  merged = pd.concat([merged, dummies_temp], axis='columns')
  merged = merged.drop([column], axis='columns')
```

```python
le = preprocessing.LabelEncoder()
le.fit(merged['nom_8'])
merged['nom_8'] = le.transform(merged['nom_8'])
```

```python
le = preprocessing.LabelEncoder()
le.fit(merged['nom_9'])
merged['nom_9'] = le.transform(merged['nom_9'])
```

```python
merged.columns
```

```python
merged['ord_1'].replace({"Novice" : 0, "Contributor" : 1, "Expert" : 2, "Master" : 3, "Grandmaster" : 4}, inplace=True)
merged['ord_2'].replace({"Freezing" : 5, "Cold" : 4, "Warm" : 3, "Hot" : 2, "Boiling Hot" : 1, "Lava Hot" : 0}, inplace=True)
```

```python
le.fit(merged['ord_3'])
merged['ord_3'] = le.transform(merged['ord_3'])
```

```python
le.fit(merged['ord_4'])
merged['ord_4'] = le.transform(merged['ord_4'])
```

```python
le.fit(merged['ord_5'])
merged['ord_5'] = le.transform(merged['ord_5'])
```

```python
dummies_temp_day = pd.get_dummies(df_train.day)
merged = pd.concat([merged, dummies_temp_day], axis='columns')
merged = merged.drop(['day'], axis='columns')
```

```python
dummies_temp_month = pd.get_dummies(df_train.month)
merged = pd.concat([merged, dummies_temp_month], axis='columns')
merged = merged.drop(['month'], axis='columns')
```

```python
merged.columns
```

```python
features = merged.drop(['id', 'target'], axis='columns')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, merged['target'], test_size=0.3, random_state=0)
```

```python
logreg = LogisticRegression( max_iter=1000, solver='lbfgs', n_jobs=-1)
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logreg.score(X_test, y_test)))
```
