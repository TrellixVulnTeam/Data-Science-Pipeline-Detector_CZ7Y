```python
test_id = test['id']
train.drop(['id'], axis=1, inplace=True)
test.drop(['id'], axis=1, inplace=True)
```

```python
print(train.shape)
print(test.shape)
ntrain = train.shape[0]
ntest = test.shape[0]
all_data = pd.concat([train, test])
print(all_data.shape)
```

```python
all_data.head()
```

```python
le = LabelEncoder()
for i in all_data.columns:
    all_data[i] = le.fit_transform(all_data[i])

print(all_data.shape)
all_data.head()
```

```python
train = all_data[:ntrain]
test = all_data[ntrain:]
print(train.shape)
print(test.shape)
```

```python
target = train['target']
train.drop(['target'], axis=1, inplace=True)
test.drop(['target'], axis=1, inplace=True)
print(train.shape)
print(test.shape)
```

```python
# xgboost
xg_model = xgb.XGBClassifier(n_estimators=200, max_depth=10, learning_rate=0.2)
xg_model.fit(train, target)
y_pred_xg_train = xg_model.predict(train)
print(accuracy_score(y_pred_xg_train, target))
```

```python
y_pred_xg = xg_model.predict_proba(test)
y_pred_xg
```

```python
sub = pd.DataFrame()
sub['id'] = test_id
sub['target'] = y_pred_xg[:, 1]
sub.to_csv('submission_xgb.csv', index=False)

# # acc on test = 76.029
```
