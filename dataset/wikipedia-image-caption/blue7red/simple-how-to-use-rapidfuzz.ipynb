# rapidfuzz process module

The process module makes it compare strings to lists of strings. This is generally more performant than using the scorers directly from Python. Here are some examples on the usage of processors in RapidFuzz:

# Example1

```python
from rapidfuzz import process, fuzz
choices = ["Atlanta Falcons", "New York Jets", "New York Giants", "Dallas Cowboys"]
```

```python
process.extract("new york jets", choices, scorer=fuzz.WRatio, limit=2)
```

```python
process.extractOne("cowboys", choices, scorer=fuzz.WRatio)
```

# Example2

```python
import pandas as pd
captions = pd.read_csv('../input/wikipedia-image-caption/test_caption_list.csv')
print(len(captions))
captions.head()
```

```python
CAPTIONS = captions.caption_title_and_reference_description.values.tolist()
print(len(CAPTIONS))
CAPTIONS[:10]
```

```python
from rapidfuzz import process, fuzz
```

```python
s = "Albert Pike Albert Pike"
```

```python
result = process.extract(s, CAPTIONS, scorer=fuzz.ratio, processor=None, limit=1)
```
