```python
#Code by https://stackoverflow.com/questions/40911170/python-how-to-read-an-image-from-a-url

import urllib.request

URL = 'https://upload.wikimedia.org/wikipedia/commons/e/e1/Thermopylae_ancient_coastline_large.jpg'

with urllib.request.urlopen(URL) as url:
    with open('temp.jpg', 'wb') as f:
        f.write(url.read())

img = PIL.Image.open('Thermopylae_ancient_coastline_large.jpg')

img.show()
```

```python
#Code by  https://stackoverflow.com/questions/7391945/how-do-i-read-image-data-from-a-url-in-python

from PIL import Image
import urllib.request
import io

URL = 'https://upload.wikimedia.org/wikipedia/commons/e/e1/Thermopylae_ancient_coastline_large.jpg'

with urllib.request.urlopen(URL) as url:
    f = io.BytesIO(url.read())

img = Image.open('Thermopylae_ancient_coastline_large.jpg')

img.show()
```
