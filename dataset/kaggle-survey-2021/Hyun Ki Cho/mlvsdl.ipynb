```python
responses_df_2021 = resp_2021

def count_then_return_percent_for_multiple_column_questions(dataframe,list_of_columns_for_a_single_question,dictionary_of_counts_for_a_single_question):
    '''
    A helper function to convert counts to percentages.
    '''
    df = dataframe
    subset = list_of_columns_for_a_single_question
    df = df[subset]
    df = df.dropna(how='all')
    total_count = len(df) 
    dictionary = dictionary_of_counts_for_a_single_question
    for i in dictionary:
        dictionary[i] = round(float(dictionary[i]*100/total_count),1)
    return dictionary


q17_list_of_columns_2021 = ['Q17_Part_1',
                       'Q17_Part_2',
                       'Q17_Part_3',
                       'Q17_Part_4',
                       'Q17_Part_5',
                       'Q17_Part_6',
                       'Q17_Part_7',
                       'Q17_Part_8',
                       'Q17_Part_9',
                       'Q17_Part_10',
                       'Q17_Part_11',
                       'Q17_OTHER']

q17_dictionary_of_counts_2021 = {
    'Linear or Logistic Regression' : (responses_df_2021['Q17_Part_1'].count()),
    'Decision Trees or Random Forests': (responses_df_2021['Q17_Part_2'].count()),
    'Gradient Boosting Machines (xgboost, lightgbm, etc)' : (responses_df_2021['Q17_Part_3'].count()),
    'Dense Neural Networks (MLPs, etc)' : (responses_df_2021['Q17_Part_6'].count()),
    'Convolutional Neural Networks' : (responses_df_2021['Q17_Part_7'].count()),
    'Recurrent Neural Networks' : (responses_df_2021['Q17_Part_9'].count()),
    'Transformer Networks (BERT, gpt-3, etc)' : (responses_df_2021['Q17_Part_10'].count()),
}



q17_dictionary_of_perc_2021 = count_then_return_percent_for_multiple_column_questions(responses_df_2021,
                                                  q17_list_of_columns_2021,
                                                  q17_dictionary_of_counts_2021)
```

```python
def count_then_return_percent(dataframe,column_name):
    '''
    A helper function to return value counts as percentages.
    '''
    counts = dataframe[column_name].value_counts(dropna=True)
    percentages = round(counts*100/(dataframe[column_name].count()),1)
    return percentages

#responses_df_2019 = responses_df_2019['Q11'].replace(["$0 (USD)"], "$0",inplace=True)

def cloud_cost_scatterplot(perc_dict, year):
    df = perc_dict.to_frame(name = 'Percentage of Respondents').iloc[1:,:]
    df = df.reset_index()
    df = df.rename(columns={'index': 'Money Spent on Cloud(USD)'})
    df['Year'] = year
    
    return df

cloud_cost_perc_2019 = count_then_return_percent(responses_df_2019,'Q11').iloc[::-1]#[responses_in_order]
cloud_cost_perc_2020 = count_then_return_percent(responses_df_2020,'Q25').iloc[::-1]#[responses_in_order]
cloud_cost_perc_2021 = count_then_return_percent(responses_df_2021,'Q26').iloc[::-1]#[responses_in_order]

cloud_cost_df = cloud_cost_scatterplot(cloud_cost_perc_2021, 2021)
cloud_cost_df = cloud_cost_df.append(cloud_cost_scatterplot(cloud_cost_perc_2020, 2020))
cloud_cost_df = cloud_cost_df.append(cloud_cost_scatterplot(cloud_cost_perc_2019, 2019))
cloud_cost_df = cloud_cost_df.reset_index(drop=True)
```
