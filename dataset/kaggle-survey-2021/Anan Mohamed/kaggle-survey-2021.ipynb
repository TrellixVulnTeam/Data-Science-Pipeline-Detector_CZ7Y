```python
#Which of the following hosted notebook products do you use on a regular basis?  (Select all that apply) - Selected Choice - Azure Notebooks
#First i put all parts for question 10 in the same list
q10 = [
    'Q10_Part_1', 'Q10_Part_2', 'Q10_Part_3', 'Q10_Part_4', 
    'Q10_Part_5', 'Q10_Part_6', 'Q10_Part_7', 'Q10_Part_8', 
    'Q10_Part_9', 'Q10_Part_10', 'Q10_Part_11', 
    'Q10_Part_12', 'Q10_Part_13','Q10_Part_14','Q10_Part_15','Q10_Part_16', 'Q10_OTHER'
]
#Second i count column for answer question part1
notebook = dataframe2['Q10_Part_1'].value_counts().reset_index()
notebook.columns = [
    'notebook', 
    'count'
]
#i used for loop for get all choice in part 10 
for i in range(1, len(q10)):
    other_choice = dataframe2[q10[i]].value_counts().reset_index()
    other_choice.columns = [
        'notebook', 
        'count'
    ]
    notebook = pd.concat([notebook, other_choice])
    notebook = notebook.sort_values(['count'])
```

```python
notebook
```

```python
f, ax = plt.subplots(figsize=(7, 3))
sns.barplot(y="notebook",x='count', data=notebook, color="b")
plt.show()
```
