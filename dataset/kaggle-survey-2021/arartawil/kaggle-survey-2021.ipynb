```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import plotly.express as px
# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

df = pd.read_csv('/kaggle/input/kaggle-survey-2021/kaggle_survey_2021_responses.csv', skiprows=1)
```

# **# In this notebook you will see analysis for kaggle survey in 2021, where analysis topic in age , country , programming language, Gender.**

```python
df = df.rename(columns = {'What is your gender? - Selected Choice': 'Gender',
                         "What is your age (# years)?":"age"
                         ,"Duration (in seconds)":"Duration",
                          "What is the highest level of formal education that you have attained or plan to attain within the next 2 years?":"education",
                          "What programming languages do you use on a regular basis? (Select all that apply) - Selected Choice - Python":"python",
                          "What programming languages do you use on a regular basis? (Select all that apply) - Selected Choice - R":"R",
                          "What programming languages do you use on a regular basis? (Select all that apply) - Selected Choice - SQL":"SQL",
                          "What programming languages do you use on a regular basis? (Select all that apply) - Selected Choice - Java":"Java"
                          ,
                          "What programming languages do you use on a regular basis? (Select all that apply) - Selected Choice - C++":"C++",
                          "What programming languages do you use on a regular basis? (Select all that apply) - Selected Choice - C":"C",
                          "What programming languages do you use on a regular basis? (Select all that apply) - Selected Choice - Javascript":"Javascript",
                           "What programming languages do you use on a regular basis? (Select all that apply) - Selected Choice - MATLAB":"MATLAB",
                          "In which country do you currently reside?":"country"
                         }, inplace = False)

data_need=df[['Duration','age','Gender','country','education','python',"R","SQL","Java","C++","C","Javascript","MATLAB"]]
data_need.head(10)
```

# ** AGE Kaggle User**

```python
import plotly
plotly.offline.init_notebook_mode(connected=True)
import plotly.graph_objs as graph

num_age = data_need['age'].value_counts().sort_values(ascending=False)

data = [graph.Bar(
        x = num_age.index, # index = category name
        y = num_age.values, # value = count
)]



plotly.offline.iplot(data)
```

```python
def country_avg(name):
    return len(data_need[data_need['country'] ==name]) /len(data_need['country'])
```

# Country

```python
country_uniqe=data_need['country'].unique();
lst_avg=[]
lst_name=[]

for co in country_uniqe:
    lst_name.append(co)
    lst_avg.append(country_avg(co))
    
Avg_country = pd.DataFrame({"Name":lst_name,"Avg":lst_avg})  
Avg_country=Avg_country.head(20)
fig = px.pie(Avg_country, values='Avg', names='Name',title="Country For each use Kaggle")
fig.show()
```

# Language Programming Use

```python
lst_prog_name=[]
lst_prog_avg=[]
python_prog=len(data_need[data_need['python']== "Python"])/len(data_need['python']);
lst_prog_avg.append(python_prog)
lst_prog_name.append("Python")

Java_prog=len(data_need[data_need['Java']== "Java"])/len(data_need['Java']);
lst_prog_avg.append(Java_prog)
lst_prog_name.append("Java")


SQL_prog=len(data_need[data_need['SQL']== "SQL"])/len(data_need['SQL']);
lst_prog_avg.append(SQL_prog)
lst_prog_name.append("SQL")

R_prog=len(data_need[data_need['R']== "R"])/len(data_need['R']);
lst_prog_avg.append(R_prog)
lst_prog_name.append("R")

Javascript_prog=len(data_need[data_need['Javascript']== "Javascript"])/len(data_need['Javascript']);
lst_prog_avg.append(Javascript_prog)
lst_prog_name.append("Javascript")

C_prog=len(data_need[data_need['C']== "C"])/len(data_need['C']);
lst_prog_avg.append(C_prog)
lst_prog_name.append("C")

CCC_prog=len(data_need[data_need['C++']== "C++"])/len(data_need['C++']);
lst_prog_avg.append(CCC_prog)
lst_prog_name.append("C++")


MATLAB_prog=len(data_need[data_need['MATLAB']== "CMATLAB"])/len(data_need['MATLAB']);
lst_prog_avg.append(MATLAB_prog)
lst_prog_name.append("MATLAB")



Avg_prog = pd.DataFrame({"Name":lst_prog_name,"Avg":lst_prog_avg})  
Avg_prog.head(10)
```

```python
fig = px.pie(Avg_prog, values='Avg', names='Name',title="Language For each use Kaggle")
fig.show()
```
