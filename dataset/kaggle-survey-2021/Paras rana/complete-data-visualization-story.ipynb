```python
def func4(df,s):
    l = list(df[s])
    l1 = list(set(l))
    l2 = []
    for i in range(len(l1)): 
        temp = ((l.count(l1[i]))*100)/len(l)
        l2.append(temp)
    return l1,l2
```

```python
df['Q2'] = df['Q2'].replace(['Man', 'Woman', 'Nonbinary', 'Prefer not to say','Prefer to self-describe'],
                 ['male','female','others','others','others'])
fig = go.Figure(data=[go.Pie(labels=func4(df,'Q2')[0], values=func4(df,'Q2')[1], hole=.3)])
fig.show()
```

```python

df2 = pd.concat([pd.DataFrame(func4(df,'Q3')[0]),pd.DataFrame(func4(df,'Q3')[1])],axis=1)
df2.columns = ['country','percent']
df2['country'] = df2['country'].replace(['United Kingdom of Great Britain and Northern Ireland','Other','I do not wish to disclose my location','United States of America','South Korea','Hong Kong (S.A.R.)','Viet Nam','Iran, Islamic Republic of...'],
                                        ['United Kingdom','other','other','united states','korea, south','hong kong','vietnam','iran'])

df2 = df2[df2['country']!='other']
def func5(s):
    df3 = pd.read_csv('../input/plotlycountrycodes/plotly_countries_and_codes.csv')
    l1 = list(df3['COUNTRY'])
    l2 = list(df3['CODE'])
    lx = [[l1[i],l2[i]] for i in range(len(l1))]
    for j in range(len(lx)):
        if (s).lower()==(lx[j][0]).lower():
            return lx[j][1]
        
df2['country_code'] = df2['country'].apply(lambda x:func5(x))
```

```python
fig = go.Figure(data=go.Choropleth(
    locations = df2['country_code'],
    z = df2['percent'],
    text = df2['country'],
    colorscale = 'Blues',
    autocolorscale=False,
    reversescale=True,
    marker_line_color='darkgray',
    marker_line_width=0.5,
    colorbar_title = 'Nation wise<br>Kagglers',
))
fig.update_layout(
    title_text='World led by Kagglers',
    geo=dict(
        showframe=False,
        showcoastlines=False,
        projection_type='equirectangular'
    ),
    annotations = [dict(
        x=0.55,
        y=0.1,
        xref='paper',
        yref='paper',
        text='KAGGLERS',
        showarrow = False
    )]
)

fig.show()
del df2
```

```python
def func5(df,s):
    l = list(df[s])
    l1 = list(set(l))
    l2 = []
    for i in range(len(l1)): 
        l2.append(l.count(l1[i]))
    return l1,l2
```
