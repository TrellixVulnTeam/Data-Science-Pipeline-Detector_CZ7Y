```python
df=pd.read_csv("/kaggle/input/kaggle-survey-2021/kaggle_survey_2021_responses.csv",low_memory=False)
df
```

```python

m1=~pd.isnull(df["Q7_Part_1"])   #python
m2=~pd.isnull(df["Q7_Part_3"])   #sql
m3=~pd.isnull(df["Q7_Part_10"])  #bash

m4=~pd.isnull(df["Q16_Part_2"])  #tensorflow
m5=~pd.isnull(df["Q16_Part_3"])  #keras

m6=~pd.isnull(df["Q27_A_Part_1"])#AWS

tot=m1.count()
```

# Next we will plot how big portion fulfills each requirement

```python
fig=px.histogram(m1,title="used python")
fig.show()
fig=px.histogram(m2,title="used sql")
fig.show()
fig=px.histogram(m3,title="used bash")
fig.show()
fig=px.histogram(m4,title="used tensorflow")
fig.show()
fig=px.histogram(m5,title="used keras")
fig.show()
fig=px.histogram(m6,title="used AWS")
fig.show()
```

# Let's add all of these together and see how big percentage fills this criteria

```python
mask=m1&m2&m3&m4&m5&m6

print(mask.sum(),"out of",tot)
print(round(mask.sum()/tot*100,2),"%")
```

We end up having less than **1%** of the population left if we require all of these skills

# Let's see the working experience in cumulative fashion

**Note:** The way of presentation can be a bit misleading, because it starts with 97% of the population. It should be read in following way: On column 'I have never written code' has all of the population that has more experience than the label.

```python
import plotly.express as px

cumsum=0
sums=[]
x_axis=[ 'I have never written code',
  '< 1 years',
  '1-3 years',
  '3-5 years',
  '5-10 years',
  '10-20 years',
  '20+ years']

for i in x_axis:
    cumsum+=round((df["Q6"][1:]==i).sum()/(df["Q6"][1:]==i).count()*100,2)
    sums.append([i,100-cumsum])
    
px.bar(pd.DataFrame(sums,columns=["experience above level","percentage %"]),x="experience above level",y="percentage %",title="Cumulative amount of workers")
```
