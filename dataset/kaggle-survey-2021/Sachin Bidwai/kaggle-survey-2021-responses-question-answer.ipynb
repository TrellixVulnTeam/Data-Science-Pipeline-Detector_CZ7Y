```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.graph_objects as go 
import matplotlib.cm as cm
from matplotlib.colors import rgb2hex




df = pd.read_csv("../input/kaggle-survey-2021/kaggle_survey_2021_responses.csv")
df.head()
```

```python
questions = df.iloc[0, :]
questions
```

```python
df = df.iloc[1:, :].reset_index(drop = True)
df.head()
```

# Q1 : What is your age (# years)?
# Answer : [List of Values]

```python
df = df.sort_values(by = 'Q1', ascending = True)
age_list = df.Q1.unique()
df.Q1.nunique()
print("Q1 : What is your age (# years)? : \n Answer : [List of Age Values in Classes ] : ",age_list)
print("Number of Classes in age_list : ",len(age_list))
```

```python
q1_vcount = df['Q1'].value_counts()
q1_graph = go.Figure()
q1_graph.add_trace(go.Bar(x = q1_vcount.index, y = q1_vcount.values))
q1_graph.show()
```

# Q2. What is your gender? 

```python
gender_list = list(set(df.Q2))
print("Q2 :  What is your gender? : \n Answer : [List of genders ] : \n  ",gender_list)
print("Number of Genders : ",len(gender_list))
```

```python
cmap = cm.get_cmap('coolwarm_r',12) 
col_def =[]
for i in range(cmap.N):
    rgb = cmap(i)[:3]
    col_def.append(rgb2hex(rgb))
   
fig, axes = plt.subplots(1,2, figsize=(12, 6))
q2_vcount = df['Q2'].value_counts()
sns.barplot(x = q2_vcount.values, y= q2_vcount.index, palette=col_def,
            edgecolor = "black",ax=axes[0])
axes[0].set_title("Gender", fontsize=20)
axes[1].pie(x= q2_vcount,labels = q2_vcount.index, colors=col_def,autopct='%.0f%%',
           explode=[0.02, 0.05, 0.2, 0.6, 1.0])
axes[1].add_artist(plt.Circle((0,0),0.4,fc='white'))
plt.show()
```
