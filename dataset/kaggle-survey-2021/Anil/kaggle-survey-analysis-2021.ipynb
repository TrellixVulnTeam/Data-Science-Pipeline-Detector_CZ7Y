
# Annual Kaggle survey analysis 2021



```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../input/kaggle-survey-2021/kaggle_survey_2021_responses.csv')
```

```python
df.head()
```

```python

# drop first row
df.drop(0,inplace = True)
```

### Kagglers country of residence(Top 20)

```python
plt.figure(figsize = (12,8))
ax = plt.axes()
ax.grid(False)
ax.set(facecolor = 'grey')
sns.countplot(x = df['Q3'],order=df.Q3.value_counts().iloc[:20].index)
plt.title('Country of Residence (Top 20)',weight = 'bold',fontsize = 18)
plt.ylabel('Total Responders',weight = 'bold',fontsize = 12)
plt.xticks(rotation = 90,weight = 'bold',fontsize = 12)
plt.yticks(weight = 'bold',fontsize = 12)
plt.show()
```

# What are the age groups of Kagglers

```python
age = df['Q1']
plt.figure(figsize = (12,8))
ax = plt.axes()
ax.grid(False)
ax.set(facecolor = 'grey')
sns.countplot(y = age,order=df.Q1.value_counts().index)
plt.title('Age group of Kagglers',weight = 'bold',fontsize = 18)
plt.xlabel('Total Responders',weight = 'bold',fontsize = 12)
plt.ylabel('Age Group',weight = 'bold',fontsize = 12)
plt.xticks(weight = 'bold',fontsize = 12)
plt.yticks(weight = 'bold',fontsize = 12)
plt.show()
```

# Education

```python
plt.figure(figsize = (12,8))
ax = plt.axes()
ax.grid(False)
ax.set(facecolor = 'grey')
sns.countplot(x = df['Q4'],order=df.Q4.value_counts().index)
plt.title('Education',weight = 'bold',fontsize = 18)
plt.ylabel('Total Responders',weight = 'bold',fontsize = 12)
plt.xticks(rotation = 90,weight = 'bold',fontsize = 12)
plt.yticks(weight = 'bold',fontsize = 12)
plt.show()
```

# Kagglers Gender

```python
plt.figure(figsize = (8,5))
labels= ['Man','Woman','Prefer not to say','Prefer to self-describe','Nonbinary']
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral','magenta']
gender = df['Q2'].value_counts()
plt.pie(gender,colors = colors, autopct='%1.1f%%', shadow=True,radius = 2,pctdistance = 1.1)
plt.legend(labels,loc = 'best')
plt.show()
```

# Education by Gender

```python
plt.figure(figsize = (12,8))
ax = plt.axes()
ax.grid(False)
ax.set(facecolor = 'grey')
sns.countplot(x = 'Q4',hue ='Q2',data= df)
plt.xticks(rotation = 90,weight = 'bold',fontsize = 12)
plt.yticks(weight = 'bold',fontsize = 12)
plt.ylabel('Total Responders',weight = 'bold',fontsize = 12)
plt.show()
```

# Job Profession

```python
plt.figure(figsize = (12,8))
ax = plt.axes()
ax.grid(False)
ax.set(facecolor = 'grey')

job_count = df['Q5'].value_counts()
total = df.shape[0]

#fig,ax = plt.subplots(figsize=[12,10])

ax.bar(job_count.index, job_count, width = 0.5)

for j in job_count.index:
       ax.annotate('{:.2f}%'.format(job_count[j]/total*100),
                  xy=(j, job_count[j]+10))

plt.xticks(rotation = 90,weight = 'bold',fontsize = 12)
plt.yticks(weight = 'bold',fontsize = 12)
plt.ylabel('No of Responders',weight = 'bold',fontsize = 12)
plt.title('Job Profession', weight = 'bold',fontsize = 18)
plt.show()

```
