```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("../input/kaggle-survey-2021/kaggle_survey_2021_responses.csv")
```

```python
df.head()
```

```python
new_header = df.iloc[0]
df = df[1:]
df.columns = new_header
```

```python
sns.countplot(x="What is your age (# years)?",hue='What is your gender? - Selected Choice', data=df)
```

```python
df['What is your age (# years)?'].value_counts().sort_index().plot.bar(
    figsize=(12, 6),
    color='mediumvioletred',
    fontsize=16
)
```

```python
df1=df['In which country do you currently reside?'].value_counts().nlargest(10)
```

```python
df1.plot.bar(
    figsize=(12, 6),
    color='mediumvioletred',
    fontsize=16
)
```

```python
df2=df['What is the highest level of formal education that you have attained or plan to attain within the next 2 years?'].value_counts().nlargest(15)
df2.plot.bar(
    figsize=(12, 6),
    color='mediumvioletred',
    fontsize=16
)
```

```python
df3=df['What programming language would you recommend an aspiring data scientist to learn first? - Selected Choice'].value_counts()
df3.plot.bar(
    figsize=(12, 6),
    color='mediumvioletred',
    fontsize=16
)
```

```python
df4=df['Which of the following business intelligence tools do you use most often? - Selected Choice'].value_counts()
df4.plot.bar(
    figsize=(12, 6),
    color='mediumvioletred',
    fontsize=16
)
```
