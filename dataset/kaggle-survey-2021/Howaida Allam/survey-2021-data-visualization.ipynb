## What is your gender?Â¶

```python
gender = data.loc[1:, "Q2"].value_counts()
categories = list(gender.keys())
palette = sns.color_palette("Set2", len(categories)) 
colors1 = {k:v for k,v in zip(categories, palette)}
colors2 = {k:v for k,v in zip(categories[:2] + ["Others"], palette[:2] + [(1, 0, 0)])}
colors3 = {k:v for k,v in zip(categories[2:], palette[2:])}
```

```python
gen1 = gender.to_frame()
gen1.loc["others"] =  gen1[2:].sum()
gen1 = gen1.drop(categories[2:])
gen1["Q2"] = gen1["Q2"]/gen1["Q2"].sum() * 100

gen2 = gender.to_frame()
gen2["Q2"] = gen2["Q2"]/gen2["Q2"].sum() * 100
gen2 = gen2.drop(categories[:2])
```

```python
# make figure and assign axis objects
fig = plt.figure(figsize = (15, 7))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
fig.subplots_adjust(wspace = 0)

# pie chart parameters
ratios = gen1["Q2"].values
labels = colors2.keys() 
explode = [0, 0, 0.2]
angle = ratios[-1]

ax1.pie(ratios, autopct = "%1.2f%%",  startangle = angle, labels = labels, 
        explode = explode, colors = colors2.values(), pctdistance = 0.85,
        wedgeprops = dict(width = 0.55))
ax1.legend(loc='upper left')

# bar chart parameters
xpos = 0
bottom = 0
ratios = gen2["Q2"].values
width = 0.2

for j in range(len(ratios)):
    height = ratios[j]
    ax2.bar(xpos, height, width, bottom = bottom, color = palette[2:][j])
    ypos = bottom + ax2.patches[j].get_height() / 2
    bottom += height
    ax2.text(xpos, ypos, "%.2f%%" % (ax2.patches[j].get_height()) + "\n" + categories[2:][j],
             va="center" ,ha  = "center")

ax2.set_title("Others",font='Comic Sans MS',fontsize=20)
ax2.axis('off')
ax2.set_xlim(- 2.5 * width, 2.5 * width)

# use ConnectionPatch to draw lines between the two plots
theta1, theta2 = ax1.patches[-1].theta1, ax1.patches[-1].theta2
center, r = ax1.patches[-1].center, ax1.patches[-1].r
bar_height = sum([item.get_height() for item in ax2.patches])

# draw top connecting line
x = r * np.cos(np.pi / 180 * theta2) + center[0]
y = r * np.sin(np.pi / 180 * theta2) + center[1]

con = ConnectionPatch( xyA = (-width / 2, bar_height), coordsA = ax2.transData, 
                       xyB = (x, y), coordsB = ax1.transData)
con.set_color([0, 0, 0])
con.set_linewidth(0.9)
ax2.add_artist(con)

# draw bottom connecting line
x = r * np.cos(np.pi / 180 * theta1) + center[0]
y = r * np.sin(np.pi / 180 * theta1) + center[1]

con = ConnectionPatch( xyA = (-width / 2, 0),  coordsA = ax2.transData,
                    xyB = (x, y), coordsB = ax1.transData)
con.set_color([0, 0, 0])
ax2.add_artist(con)
con.set_linewidth(0.9)

plt.show()
```

## Which categories of computer vision methods do you use on a regular basis? (Select all that apply)

```python
cv_cols = ['Q18_Part_1', 'Q18_Part_2', 'Q18_Part_3', 'Q18_Part_4', 'Q18_Part_5', 'Q18_Part_6', 'Q18_OTHER']

df = data.loc[1:, ["Q2"] + cv_cols]

cv = ["General purpose", "Segmentation", "Object detection", "Classification", "Generative", "None", "Other"]
df.columns = ["gender"] + cv

df[df[cv].notnull()] = 1
df = df.fillna(0)
    
df = df.groupby(["gender"]).sum()
    
df = df.apply(lambda x: x/x.sum() * 100, axis = 1)

df.columns.name = "cv"

df = df.unstack().reset_index(name = "count")

df["gender"] = pd.Categorical(df["gender"],categories = categories,ordered = True)

df = df.sort_values(["cv", "gender"])
```

```python
for c in set(df["cv"]):
    temp = df[df["cv"] == c]

    labels = list(temp["gender"])
    values = list(temp["count"])
    
    #number of data points
    n = len(values)
    #find max value for full ring
    k = 10 ** int(log10(max(values)))
    m = k * (1 + max(values) // k)

    #radius of donut chart
    r = 1.5
    #calculate width of each ring
    w = r / n 

    #create figure, axis
    fig, ax = plt.subplots()
    ax.axis("equal")

    #create rings of donut chart
    for i in range(n):
        #hide labels in segments with textprops: alpha = 0 - transparent, alpha = 1 - visible
        innerring, _ = ax.pie([m - values[i], values[i]], radius = r - i * w,  startangle = 90, labels = ["", labels[i]], 
        labeldistance = 1 - 1 / (1.5 * (n - i)),textprops = {"alpha": 0}, colors = ["white", palette[i]])
        plt.setp(innerring, width = w, edgecolor = "white")

    plt.legend(loc = "upper center")
    plt.title(c,font='Comic Sans MS',fontsize=15)
    plt.ylim(0, 3)
    plt.show()
```
