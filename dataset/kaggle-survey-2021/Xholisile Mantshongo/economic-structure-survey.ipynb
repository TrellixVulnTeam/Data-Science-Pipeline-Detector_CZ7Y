# DATA CLEANING

```python
survey.isnull().sum()>12543
```

Removing all entries with more than half the cases havin missing data

```python
n = []
for i in survey:
    if survey[i].isnull().sum() > 12987.0:
        n.append(i)
```

```python
len(n)
```

```python
survey_clean = survey.drop(n,axis=1)
```

```python
survey_clean
```

# MULTI-TABLE ANALYSIS OF THE CLEAN VERSION OF THE SURVEY 

```python
from IPython.core.display import HTML

def multi_table(table_list):
    ''' Accepts a list of IpyTable objects and returns a table which contains each IpyTable in a cell
    '''
    return HTML(
        '<table><tr style="background-color:white;">' + 
        ''.join(['<td>' + table._repr_html_() + '</td>' for table in table_list]) +
        '</tr></table>')
```

```python
survey_nunique = {var: pd.DataFrame(survey_clean[var].value_counts()) 
              for var in {'Time from Start to Finish (seconds)', 'Q1', 'Q2', 'Q3', 'Q4', 'Q5',
       'Q6', 'Q7_Part_1', 'Q8', 'Q9_Part_11', 'Q11', 'Q12_Part_5', 'Q13',
       'Q14_Part_1', 'Q15', 'Q16_Part_1', 'Q17_Part_1', 'Q20', 'Q21', 'Q22',
       'Q23', 'Q25', 'Q26', 'Q41'}}
multi_table([survey_nunique['Time from Start to Finish (seconds)'], survey_nunique['Q1'],survey_nunique['Q2']
            ,survey_nunique['Q3'],survey_nunique['Q4'],survey_nunique['Q5'],
             survey_nunique['Q6'],survey_nunique['Q7_Part_1']
            ,survey_nunique['Q8'],survey_nunique['Q9_Part_11'],survey_nunique['Q11']
            ,survey_nunique['Q12_Part_5'],survey_nunique['Q13'],survey_nunique['Q14_Part_1']
            ,survey_nunique['Q15'],survey_nunique['Q16_Part_1'],survey_nunique['Q17_Part_1']
            ,survey_nunique['Q20'],survey_nunique['Q21'],survey_nunique['Q22']
            ,survey_nunique['Q23'],survey_nunique['Q25'],survey_nunique['Q26'],survey_nunique['Q41']])
```
