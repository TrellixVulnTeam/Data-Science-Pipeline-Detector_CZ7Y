```python
fields = ['-', 'Time from Start to Finish (seconds)', 'Q1', 'Q2', 'Q3', 'Q4', 'Q15', 'Q20', 'Q21', 'Q25', 'Q26']
```

```python
kagglers_profile = pd.read_csv('/kaggle/input/kaggle-survey-2021/kaggle_survey_2021_responses.csv', usecols=lambda col: col.startswith('Q7_') or col.startswith('Q16_') or col.startswith('Q17_') or col.startswith('Q40_') or col in fields)
```

```python
kagglers_history = pd.read_csv('/kaggle/input/kaggle-data-science-survey-20172021/kaggle_survey_2017_2021.csv', usecols=lambda col: col.startswith('Q7_') or col.startswith('Q16_') or col.startswith('Q17_') or col.startswith('Q40_') or col in fields)
```

```python
kagglers_profile = kagglers_profile.drop([0])
```

```python
age_profile = kagglers_profile['Q1'].value_counts()

fig = go.Figure()
fig.add_trace(go.Bar(x = age_profile.index, y = age_profile.values, marker_color = '#FFA15A'))

fig.update_layout(plot_bgcolor = "white",
                 font  = dict(color = "#636EFA"), 
                 title = dict(text = "Kagglers Age"), 
                 xaxis = dict(title = "Age Range", linecolor = "#21DBAA"), 
                 yaxis = dict(title = "Count", linecolor = "#DB9021"))
fig.show()
```

```python
kagglers_profile['Q25'] = kagglers_profile['Q25'].str.replace('$', '')
kagglers_profile['Q25'] = kagglers_profile['Q25'].str.replace('>', '')
kagglers_profile['Q25'] = kagglers_profile['Q25'].str.replace(',', '')
kagglers_profile['Q25'] = kagglers_profile['Q25'].str.rsplit('-', 1).str.get(0)
# kagglers_profile['Q25'] = kagglers_profile['Q25'].str.join("+")
```

```python
kagglers_profile['Q25'] = kagglers_profile['Q25'].astype('float')
```

```python
kagglers_profile = kagglers_profile.sort_values(['Q25'])
```

```python
kagglers_profile['Q25'].unique()
```

```python
gender = kagglers_profile['Q2'].value_counts()
gender.head()
```

```python
gender_salary = kagglers_profile.groupby(['Q2', 'Q25']).size().reset_index().rename(columns = {0:'Count'})
gender_salary['pct_by_gender'] = gender_salary['Count'] / gender_salary.groupby(['Q2'])['Count'].transform('sum') * 100
gender_salary.head()
```

```python
fig = go.Figure()
for gender, group in gender_salary[gender_salary['Q2'].isin(['Man', 'Woman'])].groupby("Q2"):
   fig.add_trace(go.Bar(x = group['Q25'].astype(str), y = group['pct_by_gender'], name = gender))
fig.update_layout(barmode="stack", 
                 plot_bgcolor = "white")
fig.show()
```

In the low pay brackets the percent of women is higher. With salary growing we can see that men prevail compared to women. 

```python
fig = go.Figure()
for gender, group in gender_salary[~gender_salary['Q2'].isin(['Man', 'Woman'])].groupby("Q2"):
   fig.add_trace(go.Bar(x = group['Q25'].astype(str), y = group['pct_by_gender'], name = gender))
fig.update_layout(barmode="stack", 
                 plot_bgcolor = "white")
fig.show()
```

```python
age_gender = kagglers_profile.groupby(['Q1', 'Q2']).size().reset_index().rename(columns = {0:'Count'})
age_gender.head()
```

```python
fig = go.Figure()
for gender, group in age_gender[age_gender['Q2'].isin(['Man', 'Woman'])].groupby("Q2"):
   fig.add_trace(go.Bar(x = group['Q1'], y = group['Count'], name = gender))
fig.update_layout(barmode="relative", 
                 plot_bgcolor = "white")
fig.show()
```
