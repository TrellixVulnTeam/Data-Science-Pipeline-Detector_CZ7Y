```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
sns.set_style('white')
warnings.filterwarnings('ignore')
```

```python
kaggle_survey=pd.read_csv("../input/kaggle-survey-2021/kaggle_survey_2021_responses.csv")
kaggle_survey = kaggle_survey.iloc[1:,:]
```

```python
continents_df = pd.read_csv("../input/2020-kaggle-survey-supporting/2020 Continents.csv")
continents_dict = {k:v for k,v in zip(continents_df["Country"],continents_df["Continents"])}
kaggle_survey["Continents"] = kaggle_survey["Q3"].map(continents_dict)
```

```python
country_long_name_dict = {"United States of America": "USA", 
                          "United Kingdom of Great Britain and Northern Ireland": "United Kingdom",}
kaggle_survey["Q3"] = kaggle_survey["Q3"].replace(country_long_name_dict)

education_categorical_dict={
    "No formal education past high school":"Below high school",
    "Some college/university study without earning a bachelorâ€™s degree":"University study w/o bachelor",
    "I prefer not to answer":"No answer",
}
kaggle_survey["Q4"] = kaggle_survey["Q4"].replace(education_categorical_dict)
```

# **1.World**

```python
continents_count_df = pd.DataFrame(kaggle_survey["Q4"].value_counts())
continents_count_df = continents_count_df.reset_index(drop=False)
continents_count_df.columns = ["Education","Count"]
```

```python
fig = plt.figure(figsize=(10,8))
color_map = ["#112A86", "#f04fc5", "#bdbdbd",'#f4acb7','#9d8189','#E63946','#8D99AE']
plt.text(0, 1.2, 'World', fontsize=18, fontweight='bold', fontfamily='serif', horizontalalignment='center')
plt.pie(x=continents_count_df['Count'], colors=color_map, wedgeprops=dict(width=0.2),autopct='%1.0f%%')
plt.legend(continents_count_df["Education"], loc="lower center", bbox_to_anchor=(-1, 1))
```

* **39% of Kagglers have Master degree followed by 38% Bachelor degree.**
* **11% of Kagglers have Doctoral degree.**
