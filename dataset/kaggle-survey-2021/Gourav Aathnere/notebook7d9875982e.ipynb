## Number of Profeesionals Participated in the Survey?

```python
df.rename(columns={'Select the title most similar to your current role (or most recent title if retired): - Selected Choice':'role'}, inplace=True)
```

```python
data_role = df.role.value_counts()
data_role
```

```python
roles = ['Student', 'Data Scientist', 'Software Engineer', 'Other', 'Data Analyst', 'Not Employed', 'Research Scientist',
        'ML Engineer', 'Business Analyst', 'Project Manager', 'Data Engineer', 'Product Manager', 'Statistician', 'DBA',
        'Developer Advocacy']
roles
```

```python
roles_count = [data_role[i] for i in range(0,15)]
roles_count
```

```python
xpos = np.arange(len(roles))
xpos
```

```python

plt.bar(xpos,roles_count, label="Profession Count", width=0.5, alpha=0.8)

plt.xticks(xpos,roles)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend();
```

```python
ax = sns.barplot(x=df.role.value_counts(), y=roles, data=df)
```
