```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import warnings; warnings.filterwarnings(action='once')

large = 22; med = 16; small = 12
params = {'axes.titlesize': large,
          'legend.fontsize': med,
          'figure.figsize': (16, 10),
          'axes.labelsize': med,
          'axes.titlesize': med,
          'xtick.labelsize': med,
          'ytick.labelsize': med,
          'figure.titlesize': large}
plt.rcParams.update(params)
plt.style.use('seaborn-whitegrid')
sns.set_style("white")
%matplotlib inline

# Version
print(mpl.__version__)  #> 3.0.0
print(sns.__version__)  #> 0.9.0
```

```python
df = pd.read_csv("/kaggle/input/kaggle-survey-2021/kaggle_survey_2021_responses.csv")
#Dropping the question text row
df.drop(0,axis=0,inplace=True)
#Converting responses to gender other than man or woman to other due to lesser count
df.at[(df['Q2'] != 'Man') & (df['Q2'] != 'Woman'),'Q2'] = 'Other'
df.shape
```

```python
#Simple demographics
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2)
fig.set_size_inches(30,30)
ax1.pie(df['Q1'].value_counts(),labels =df['Q1'].value_counts().index,autopct='%d%%',colors=reversed(sns.color_palette('mako',12)))
ax2.pie(df['Q2'].value_counts(),labels =df['Q2'].value_counts().index,autopct='%d%%',colors=reversed(sns.color_palette('mako',12)))

cdf = df[df.Q3 != 'Other']['Q3'].value_counts().sort_values(ascending=False).reset_index()
ax3.bar(cdf.iloc[0:10]['index'],cdf.iloc[0:10]['Q3'],color=plt.cm.GnBu(np.linspace(1, 0.5, 10)))
ax3.tick_params(labelrotation=90)
ax3.set_xlabel('')
ax3.set_ylabel('')
ax4.bar(df['Q4'].value_counts().index,df['Q4'].value_counts(),color=plt.cm.GnBu(np.linspace(0.5, 2, 10)))
ax4.tick_params(rotation=90)
ax4.set_xlabel('')
ax4.set_ylabel('')
plt.show()
```
