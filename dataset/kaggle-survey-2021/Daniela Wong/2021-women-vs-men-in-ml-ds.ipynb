Decision Tree

```python
#Predicting if men or women based on their university level of education(bacherlor's or master's), coding experience , and  country will have a six figures salary or not
```

```python
df['Q3'].drop(df[df['Q3'] == "I do not wish to disclose my location"]['Q3'].index, inplace = True)
```

```python
df_dt = df[['Q2','Q4','Q3','Q5','Q6','Q25']]
df_dt.head(10)
```

```python
dict = {'Q2': 'Gender','Q3':'Country','Q4': 'Education','Q5':'Job Title','Q6': 'Years of Coding','Q25': 'Salary'}
df_dt.rename(columns=dict,inplace=True)
```

```python
df_dt = df_dt[(df_dt['Gender'] == 'Man' ) | (df_dt['Gender'] == 'Woman' )]
df_dt = df_dt[(df_dt['Education'] == 'Master’s degree' ) | (df_dt['Education'] == 'Bachelor’s degree' )]
df_dt = df_dt[(df_dt['Job Title'] == 'Data Scientist' ) | (df_dt['Job Title'] == 'Data Analyst' )]
df_dt = df_dt[(df_dt['Years of Coding'] == '3-5 years' ) | (df_dt['Years of Coding'] == '5-10 years' )]
```

```python
df_dt.isnull().sum()
```

```python
X = df_dt.drop(columns=['Salary'], axis=1)
```

```python
X = X.replace({"Woman": 0, "Man": 1})
X = X.replace({"Bachelor’s degree": 0, "Master’s degree": 1})
X = X.replace({"Data Scientist": 0, "Data Analyst": 1})
X = X.replace({"3-5 years": 0, "5-10 years": 1})
```

```python
X = pd.get_dummies(X)
```

```python
def salary(value):
    if len(value) > 15:
        return 2
    elif len(value) == 15:
        return 1
    else: return 0
```

```python
y = df_dt['Salary'].apply(salary)
```

```python
from sklearn import tree
model1 = tree.DecisionTreeClassifier(criterion='entropy')
```

```python
model1.fit(X,y)
```

```python
model1.score(X,y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
DT = tree.DecisionTreeClassifier()
DT.fit(X_train, y_train)
DT_predicted_test = DT.predict(X_test)
```

```python
print(metrics.classification_report(y_test, DT_predicted_test))
```

```python
print("accuracy score:", round(accuracy_score(y_test, DT_predicted_test),2))
print("precision score:", round(precision_score(y_test, DT_predicted_test),2))
print("recall score:", round(recall_score(y_test, DT_predicted_test),2))
```
