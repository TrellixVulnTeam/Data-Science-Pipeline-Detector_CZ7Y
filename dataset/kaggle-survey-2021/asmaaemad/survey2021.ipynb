```python
import pandas as pd
import numpy as n
import matplotlib.pyplot as plt
import plotly.express as px
%matplotlib inline
import seaborn as sns
from matplotlib import pyplot as plt
import plotly.graph_objects as go
import squarify
```

```python
#reading the data
survey= pd.read_csv("/kaggle/input/kaggle-survey-2021/kaggle_survey_2021_responses.csv")
```

```python
#explore data and cleaning
survey= survey.iloc[1:, :]
survey.head(20)
```

```python
#What is your age (# years)?
age_value = survey["Q1"].value_counts().reset_index()
age_value.columns = ["age","count"]
print(age_value)
```

```python
#Visulisation

plt.style.use("dark_background")  #style

myexplode=(0.18,0.18,0.18,0.18,0.18,0.18,0.18,0.18,0.18,0.18, 0.18)
mypalette = ["#d7e1ee", "#cbd6e4", "#bfcbdb", "#b3bfd1", "#a4a2a8", 
             "#df8879", "#c86558", "#b04238", "#991f17",
             "#146356", "#6b506b"]

fig_age, ax_age = plt.subplots(figsize=(18,10)) #creating figure

age_value.plot(kind="pie", y="count", legend = True, title ="Age Distribution", ax=ax_age,
                   labeldistance=1.1, pctdistance= 0.7, shadow = True, labels=age_value["age"],
                    fontsize = 15, explode=myexplode, colors=mypalette,
                    wedgeprops = {'linewidth':0.5, 'edgecolor':'grey'},
               autopct='%1.2f%%', startangle = 192,
              ylabel="");

plt.legend(loc = "upper right", fontsize =14 ) #location of legend


fig_age.tight_layout()
```
