<h3> <span style="font-family:Georgia;color:black;font-weight:bold;background:#EBE3DB"> 2.7 What programming languages do you use on a regular basis?

```python
def ques_title(ques):
    title = questions.filter(regex = ques)[0].split("?")[0]+'?'
    return title
ques_title('Q7')
```

```python
def plot(ques,lab, fsize = (16, 9)):
    a = df.filter(regex = ques)
    answer_desc = a.describe()
    answergot = answer_desc.loc[['top', 'count']].T
    answergot = answergot.set_index('top')
    answergot = answergot.sort_values(by="count", ascending = False)
    
    plt.figure(figsize = fsize)
    A = sns.barplot(data = answergot,
                 x='count',
                 y = answergot.index,
              color=colors[2],
              edgecolor='white',
              linewidth=1.5,
              saturation=1.5)
    
    plt.xlabel('Count', weight='semibold', fontname = 'Georgia')
    plt.ylabel(lab, weight='semibold', fontname = 'Georgia')
    plt.suptitle(ques_title(ques), fontname = 'Georgia', weight='bold')
    
    A.bar_label(A.containers[0])
    
    plt.show
   
    
    
```

```python
plot('Q7','Language')
```
