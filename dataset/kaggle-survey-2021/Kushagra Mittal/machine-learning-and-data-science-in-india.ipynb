```python
def datacol(x):
    age= x.unique()
    age_grp = age[1:]
    tmp = list(x)
    cn = []
    for i in age_grp:
        cn.append(tmp.count(i))
    return age_grp, cn
age_grp, cn = datacol(data['Q1'])
```
