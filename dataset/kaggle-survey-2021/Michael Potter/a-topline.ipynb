#  Need to create a "Completed the intervew" dataset as many people do not finish and hence the bases distorted.

# To do this, we only select the respondents that answered the last question, Q42.
# After this, we have n=18,329 valid, completed surveys


```python
def getTargetCols(df,stem):
    # df=df[0:100]
    cols=[]
    for coll in df.columns:
        if coll.startswith(stem):
            cols.append(coll)
    col_index=[df.columns.get_loc(colllee) for colllee in cols]
    rowindex=[]
    for i,resp in df.iterrows():
        b=[]
        for col in col_index:
            if resp[col] != NoAnsString:
                b.append(resp[col])
        if len(b)>0:
            rowindex.append(i-1)
    return rowindex

ri=getTargetCols(df,'Q42')

completes = df.iloc[ri]

df=completes #[ : 200]
# print(len(completes))
```

# A Filter is to reduce the data set to just the respondents we are interested in

```python
def makeFilterSeries():
    FSeries=[]
    for i, resp in df.iterrows():
        if not resp['Q5'] in ['Currently not employed', 'Student']:
            FSeries.append('Professional')
        else:
            FSeries.append(NoAnsString)
    return FSeries
FilterData=makeFilterSeries()

def makeBannerSeries():
    bannercol=[]
    for i, resp in df.iterrows():
        bannerentry=['.Total']
        if resp['Q40_Part_3'] == 'Kaggle Learn Courses':
            bannerentry.append('Courses - Kaggle Learn')
        else:
            bannerentry.append('Courses - NOT Kaggle')
        if resp['Q40_Part_1'] != NoAnsString:
            bannerentry.append('Courses - Coursera')
        if resp['Q40_Part_7'] != NoAnsString:
            bannerentry.append('Courses - Udemy')
        if resp['Q3'] == 'India':
            bannerentry.append('Country - India')
        if resp['Q3'] == 'United States of America':
            bannerentry.append('Country - U.S.A')
        if resp['Q3'] != 'India' and resp['Q3'] != 'United States of America':
            bannerentry.append('Country - Rest of the World')
        if resp['Q5'] in ['Currently not employed', 'Student']: 
            bannerentry.append('Status - Student / unemployed')
        else:
            bannerentry.append('Status - Working')
        if resp['Q1'] in ['18-21', '22-24','25-29']:
            bannerentry.append('Age - Below 30 yrs')
        else:
            bannerentry.append('Age - Over 30 yrs')
        bannercol.append(bannerentry)
        # print(bannerentry)
    return bannercol
    
BannerData=makeBannerSeries()
# print(BannerData[1:25])
# print(len(EduBanner))
```

```python
def labelslots(xx:list):
    val=0
    labellist=[]
    for  r in xx:
        for k in r:
            if not k in labellist:
                labellist.append(k)
    return sorted(labellist)

# a=labelslots(BannerSeries)
# print(a)
```

```python
def PrepData(q:list,isBanner):
    targetcols=[]
    for col in df.columns:
        if [col]==q:
            targetcols.append(q[0])
        else:
            if col.startswith(q[0] + '_Part'):
                targetcols.append(col)
    ret=[]
    
    a=df[targetcols].values.tolist()
    for bp in a:
        if isBanner:
            bp.append('.Total')
        # if bp != f"['{NoAnsString}']":
        ret.append(bp)
    return ret

# res=PrepData(['Q7'])
# print(res[1:10])
```

```python
def MultiTab(side:list,B:list,F:list=['All'],colpercents=True,showbase=True):
    xtabdata = pd.DataFrame( index=df.index).fillna(NoAnsString)
    if len(B) == 1:
        C=PrepData(B,True)
        if len(C) != len(xtabdata):
            print(C)
        else:
            B= C
    xtabdata['Banner'] = B
    S=PrepData(side,False)
    xtabdata['SideData'] = S
    if len(F) ==1: #'all':
        F=list(itertools.repeat(['All'],len(xtabdata)))
    xtabdata['Filter'] = F
    FilterLabel = list(set(FilterData))
    FilterLabel.remove(NoAnsString) 
    rawcrosstab=pd.DataFrame(columns=labelslots(B), index=labelslots(S)).fillna(0).astype(float)
    bases=pd.DataFrame(columns=rawcrosstab.columns,index=['.Answered','counted']).fillna(0).astype(float)
    for rownumber, xtabdataField in xtabdata[['Banner','SideData','Filter']].iterrows():
        bases.at['counted'] = 0
        if xtabdataField['Filter'] != NoAnsString:
            for bannerentry in xtabdataField['Banner']:
                if bannerentry != NoAnsString:
                    # if bannerentry=='Udemy':
                    #     print(rownumber,'Udemy')
                    for sideentry in xtabdataField['SideData']:
                        if sideentry != NoAnsString:
                            rawcrosstab.at[sideentry,bannerentry] += 1
                            if bases.at['counted',bannerentry] == 0:
                                bases.at['.Answered',bannerentry] = 1 + bases.at['.Answered',bannerentry]
                                bases.at['counted',bannerentry] = 1
    if colpercents:
        displaytab = rawcrosstab.divide(bases.loc['.Answered'], axis='columns').astype(float)
        displaytab = displaytab.mul(100).round(1)   #.astype(str) + '%' can cause problems
    else:
        displaytab=rawcrosstab

    if showbase:
        displaytab=pd.concat([bases.loc['.Answered'].to_frame().transpose(),displaytab],axis=0).fillna(0)
    displaytab=displaytab.loc[(displaytab.T != 0).any()]
    displaytab=displaytab.loc[:,(displaytab != 0).any(axis=0)]
    displaytab.sort_index()
    return displaytab
```
