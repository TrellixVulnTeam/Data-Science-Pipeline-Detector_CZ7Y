```python
import pandas as pd
import numpy as np
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.figure_factory as ff
import plotly.offline as offline
import plotly.graph_objs as go
offline.init_notebook_mode(connected = True)
```

```python
df = pd.read_csv('../input/kaggle-survey-2021/kaggle_survey_2021_responses.csv', low_memory=False)
```

```python
df_questions = df.iloc[0,:].reset_index().rename(columns={'index': 'Question #', 0:'Question'})[1:]
df_questions.head()
```

```python
df_data = df.iloc[1:,:]
df_data.head()
```

```python
ages = pd.DataFrame({c : df_data[c].value_counts() for c in {'Q1'}}).reset_index().rename(columns={'index':'Age_Bin','Q1':'Users'})
```

```python
countries = pd.DataFrame({c : df_data[c].value_counts() for c in {'Q3'}}).reset_index().rename(columns={'index':'Countries', 'Q3':'Users'})
```

```python
fun_countries = px.funnel(countries[:10], x='Users', y='Countries',
                #title='Top 10 Popular countries in Kaggle',
                color='Countries',
                width=800, height=400)
fun_countries.update_yaxes(categoryorder='total descending')
fun_countries.update_traces(textposition='auto', textfont=dict(color='#fff'))
fun_countries.update_layout(title={'text': "Top 10 Popular countries in Kaggle",'y':0.9,'x':0.5,'xanchor': 'center','yanchor': 'top'},
                            margin=dict(t=110, b=50, l=70, r=40),
                            xaxis_title=' ', yaxis_title=" ",
                            showlegend=False,
                            title_font=dict(size=25, color='#a5a7ab',family="Lato, sans-serif"),
                            font=dict(color='#8a8d93'))
```

```python
px.choropleth (data_frame= countries, locations = 'Countries', locationmode = 'country names', color = 'Users', scope = 'world' )
```

<p style="font-size:15px; font-family:verdana; line-height: 1.7em; margin-left:20px">   
<b> What about the Kagglers from South America? (My region) </b>  </p>

```python
country = ['Argentina', 'Brazil', 'Bolivia', 'Colombia','Chile','Ecuador', 'Peru', 'Paraguay','Uruguay','Venezuela']
df_la_kagglers = countries[countries['Countries'].isin(country)]
df_la_kagglers['% of Total Kagglers'] = ((df_la_kagglers['Users'] / countries['Users'].sum()) * 100).round(2)
colors = {'Argentina': 'lightblue','Brazil':'greenyellow','Colombia': 'yellow','Peru':'orangered','Chile':'red', 'Ecuador':'limegreen'}
df_la_kagglers['color'] = df_la_kagglers['Countries'].map(colors)

fig = make_subplots(rows=1, cols=2, 
                    specs=[[{"type": "bar"}, {"type": "pie"}]],
                    column_widths=[0.5, 0.5], vertical_spacing=0, horizontal_spacing=0.02,
                    subplot_titles=("Top 5 Countries by Users in SA", "Percentage of Total Kagglers"))

fig.add_trace(go.Bar(x=df_la_kagglers['Users'], y=df_la_kagglers['Countries'][:5], marker=dict(color=df_la_kagglers['color']),
                     name='# Kagglers', orientation='h', text=df_la_kagglers['Users']), 
                     row=1, col=1)

fig.add_trace(go.Pie(values=df_la_kagglers['% of Total Kagglers'], labels=df_la_kagglers['Countries'][:5], name='% Kagglers',
                     marker=dict(colors=['greenyellow','yellow','lightblue','orangered','red']), hole=0.7,
                     hoverinfo='label', text=df_la_kagglers['% of Total Kagglers'], textinfo='label'), 
                    row=1, col=2)

fig.update_yaxes(showgrid=False, ticksuffix=' ', categoryorder='total ascending', row=1, col=1)
fig.update_yaxes(showgrid=False, ticksuffix=' ', categoryorder='total ascending', row=1, col=2)
fig.update_traces(texttemplate='%{text}', textposition='inside', row=1, col=1)
fig.update_traces(texttemplate='%{text}', textposition='inside', row=1, col=2)
fig.update_xaxes(visible=False, row=1, col=1)
fig.update_xaxes(visible=False, row=1, col=2)
fig.update_layout(#title={'text': "Kagglers users comparassion South America vs the World",'y':0.9,'x':0.5,'xanchor': 'center','yanchor': 'top'},
                  template="plotly_white",
                  title_font=dict(size=29, color='#8a8d93', family="Lato, sans-serif"),
                  font=dict(color='#8a8d93'), 
                  hoverlabel=dict(bgcolor="#f2f2f2", font_size=13, font_family="Lato, sans-serif"),
                  showlegend=False)
fig.show()
```

 ðŸ“Œ &nbsp; We can see how Kaggle it's present in all the continents, Asia it's were more Kagglers from, with India ahead. Also in Latin America Brazil it's where the mayority of the kagglers from, and my country (Argentina), it's in the thrid place.
