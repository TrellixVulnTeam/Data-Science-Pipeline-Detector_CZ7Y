```python
#importing libraries
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../input/kaggle-survey-2021/kaggle_survey_2021_responses.csv')
questions = df[0:1 ].T
data = df[1:]
```

```python
ses_uniq = data[["Q1", "Q2", "Q3", "Q4"]].value_counts()
ses_uniq.head(10)
```

```python
f, axs = plt.subplots(1,3,
                      figsize=(20,5),
                      sharey=False)
counts = data['Q1'].value_counts()/data.shape[0]*100
sns.set(style="whitegrid")
sns.barplot(x = counts.values, y=counts.index, edgecolor = "black", ax=axs[0])
#plt.title('Age distribution, %', fontsize=20)

counts = data['Q3'].value_counts()/data.shape[0]*100
sns.barplot(x = counts.values, y=counts.index, edgecolor = "black", 
            order=data.Q3.value_counts().iloc[:10].index, ax=axs[1])
#plt.title('Country of residence distribution, %', fontsize=20)

counts = data['Q4'].value_counts()/data.shape[0]*100
sns.barplot(x = counts.values, y=counts.index, edgecolor = "black", 
            ax=axs[2])
```

```python
data['Q2'] = data['Q2'].replace(['Man', 'Woman', 'Nonbinary', 'Prefer not to say','Prefer to self-describe'],
                 ['Male','Female','others','others','others'])
counts = data['Q2'].value_counts()/data.shape[0]*100
sns.barplot(x = counts.values, y=counts.index, edgecolor = "black")
```

```python
q7 = [col for col in data if col.startswith('Q7')]
data_q7 = data[q7]
data_q7.apply(pd.Series.value_counts)
df1 = data_q7.melt(var_name='columns', value_name='index')
df1
df1 = pd.crosstab(index=df1['index'], columns=df1['columns'])
df1["sum"] = df1. sum(axis=1)
language = pd.DataFrame(df1["sum"])/data.shape[0]*100
language["language"] = language.index

language.sort_values(by = ["sum"])

sns.barplot(x = "sum" , y="language", data = language , edgecolor = "black", order= language.sort_values("sum", ascending=False).language)
plt.xlabel("% of respondents", size=15)
plt.ylabel("Programming languages", size=15)
plt.title("What programming languages do you use on a regular basis?", size=18)
```

```python
q9 = [col for col in data if col.startswith('Q9')]
data_q7 = data[q9]
data_q7.apply(pd.Series.value_counts)
df1 = data_q7.melt(var_name='columns', value_name='index')
df1
df1 = pd.crosstab(index=df1['index'], columns=df1['columns'])
df1["sum"] = df1. sum(axis=1)
language = pd.DataFrame(df1["sum"])/data.shape[0]*100
language["language"] = language.index

language.sort_values(by = ["sum"])

sns.barplot(x = "sum" , y="language", data = language , edgecolor = "black", order= language.sort_values("sum", ascending=False).language)
plt.xlabel("% of respondents", size=15)
plt.ylabel("Integrated development environments", size=15)
plt.title("Which of the following IDE's do you use on a regular basis?", size=18)
```
