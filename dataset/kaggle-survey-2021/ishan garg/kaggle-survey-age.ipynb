**THIS NOTEBOOK IS BASED ON KAGGLE SURVEY 2021 COVERING INFORMATION ABOUT KAGGLE USERS WITH RESPECT TO THEIR AGE**

```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
sns.set_style('dark')
warnings.filterwarnings('ignore')
```

```python
survey= pd.read_csv("../input/kaggle-survey-2021/kaggle_survey_2021_responses.csv")
survey.head(10)
```

```python
survey_all = survey.iloc[1:,:]
survey_all.head()
```

```python
continents_df = pd.read_csv("../input/2020-kaggle-survey-supporting/2020 Continents.csv")
continents_df.head(10)
```

```python
continents_dict = d = {k:v for k,v in zip(continents_df["Country"],continents_df["Continents"])}
continents_dict
```

```python
survey["Continents"] = survey["Q3"].map(continents_dict)
survey
```

```python
survey_asia = survey[survey["Continents"]=="Asia"]
survey_america = survey[survey["Continents"]=="America"]
survey_europe = survey[survey["Continents"]=="Europe"]
survey_africa = survey[survey["Continents"]=="Africa"]
survey_australia = survey[survey["Continents"]=="Australia"]
survey_others = survey[survey["Continents"]=="Others"]
```

# World

```python
plt.figure(figsize=(15,5))
plt.text(4,5500,'World',fontsize=15,fontweight='bold',fontfamily='Arial')
color_map=('')
plt.xticks(rotation=90)
continents_count_df = pd.DataFrame(survey_all["Q1"].value_counts())
continents_count_df = continents_count_df.reset_index(drop=False)
continents_count_df.columns = ["Age","Count"]
continents_count_df = continents_count_df.sort_values(by="Age")
sns.barplot(x=continents_count_df["Age"],y=continents_count_df['Count'])
color_map
plt.text(-0.5, -1700,'Plot Shows that most of Kagglers are young people below 30 years range and above 25 years range', 
         fontsize=14, 
         fontweight='500', 
         fontfamily='arial',
        )

color_map = ["#4A4655" for _ in range(11)]
color_map[2] = "#4898EF"

plt.grid(color='black', linestyle=':', axis='y', zorder=0,  dashes=(1,5))
plt.bar(continents_count_df["Age"], continents_count_df['Count'], 
       color=color_map, zorder=3
)
```
