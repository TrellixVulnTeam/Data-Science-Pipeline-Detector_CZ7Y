```python
# importing the libraries
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
# importing the dataset
data=pd.read_csv('../input/kaggle-survey-2021/kaggle_survey_2021_responses.csv')
data.sample(5)
```

### What is your age (years)?
![](https://images.theconversation.com/files/319594/original/file-20200310-61120-1wxbcqh.jpg?ixlib=rb-1.1.0&rect=2206%2C323%2C5417%2C2460&q=45&auto=format&w=496&fit=clip)

```python
age=data['Q1'].iloc[1:].sort_values(ascending=True)
fig=px.histogram(data,x=age,color=age,template='plotly_dark')
fig.update_layout(
    title="Age group of Kaggle users",
    xaxis_title="Age group")
fig.show()
```

we can notice that most of the kagglers are youngsters and as the age starts increasing , the number of users starts decreasing

### What is your gender?


```python
gender=data['Q2'].iloc[1:]
fig=px.pie(gender,names=gender,template='plotly_dark')
fig.update_layout(
    title="Gender distribution of Kaggle users",
    xaxis_title="Gender")
fig.show()
```

We can see that about 79% users are Man and only 18.8% users are Woman

```python
# gender by age group
fig=px.histogram(data,x=age,color=gender,template='plotly_dark')
fig.update_layout(
    title="Age group of Kaggle users",
    xaxis_title="Age group")
fig.show()
```

### In which country do you currently reside?
![](https://www.i-sight.com/wp-content/uploads/2018/10/iStock-466580732-640x300.jpg)

```python
country=data['Q3'].iloc[1:]
fig=px.histogram(data,x=country,color=country,width=1600, height=800,template='plotly_dark',orientation='v')
fig.update_layout(title='Country Distribution',
                 xaxis_title='Country')
fig.show()
```

Wow, we can see that majority of users are from India followed by USA

### What is the highest level of formal education that you have attained or plan to attain within the next 2 years?
![](https://www.unicef.org/turkey/sites/unicef.org.turkey/files/styles/hero_mobile/public/DSCF0182.jpg?itok=SOjIQmJ1)

```python
ed=data['Q4'].iloc[1:]
fig = px.pie(ed, names=ed, title='Formal education of Kagglers',width=1200,height=600,template='plotly_dark')
fig.show()
```

As expected, most of the users hold a Bachelor's or Master's degree (around 10k each)

### Select the title most similar to your current role (or most recent title if retired):
![](https://previews.123rf.com/images/robuart/robuart1807/robuart180701491/105603226-professions-people-vector-icons-set-different-profession-men-and-women-cartoon-characters-in-uniform.jpg)

```python
job=data['Q5'].iloc[1:]
fig=px.histogram(data,y=job,color=job,width=1200, height=600 ,template='plotly_dark',orientation='h')
fig.update_layout(title='Profession of Kaggle users',
                 xaxis_title='Profession')
fig.show()
```

We can see that most of the users are still student (26.2%) and Profession wise most of them are Data Scientists (13.9%)

### For how many years have you been writing code and/or programming?
![](https://i.pinimg.com/736x/5e/f1/f2/5ef1f2339b766b9c15ab11432f543fa9.jpg)

```python
years=data['Q6'].iloc[1:]
fig=px.histogram(data,y=years,color=years,width=1200, height=600 ,template='plotly_dark',orientation='h')
fig.update_layout(title='Years of coding experience',
                 xaxis_title='Years')
fig.show()
```

We can see that most of us are beginners (like me) who have 0-3 years of coding experience , surprisingly there are 1032 people who have never written code.

### What type of computing platform do you use most often for your data science projects? 
![](https://images.martechadvisor.com/images/uploads/content_images/best_cloud_computing_platforms_for_midsized_businesses_5c473d795ab87.jpg)

```python
platform=data['Q11'].iloc[1:].dropna()
fig=px.histogram(data,x=platform,color=platform,width=1200, height=600,template='plotly_dark',orientation='v')
fig.update_layout(title='Coding Platform',
                 xaxis_title='Platform')
fig.show()
```

Most of us prefer laptop and Personal computers to code

### 'In what industry is your current employer/contract (or your most recent employer if retired)?'
![](https://images.livemint.com/rf/Image-621x414/LiveMint/Period1/2016/01/04/Photos/sector-web-km7G--621x414@LiveMint.jpg)

```python
emp=data['Q20'].iloc[1:].dropna()
fig=px.histogram(data,x=emp,color=emp,width=1200, height=600,template='plotly_dark',orientation='v')
fig.update_layout(title='Employer/Contract Status',
                 xaxis_title='Employment Industry')
fig.show()
```

As expected , most of the users are from Computer/Technology industry and Academics/Education since most of us are students

### 'What is your current yearly compensation (approximate $USD)?
![](https://assets-global.website-files.com/5b5aa355afe474a8b1329a37/5fc4f47c4398749f82863b3b_comp-strat.jpg)

```python
comp=data['Q25'].iloc[1:].dropna()
fig=px.histogram(data,y=comp,color=comp,width=1200, height=600,template='plotly_dark',orientation='h')
fig.update_layout(title='Current Yearly Compensation (USD)',
                 xaxis_title='Compensation')
fig.show()
```
