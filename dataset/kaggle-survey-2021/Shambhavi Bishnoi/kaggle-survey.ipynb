```python
df["Q16_Part_1"]
```

```python
Q16={}
for i in range(1,18):
    Q16.update(dict(df[f"Q16_Part_{i}"].value_counts()))
Q16=pd.DataFrame(Q16.items(),columns=["Frameworks","count"])    
    
    
```

```python
import matplotlib.pyplot as plt
```

```python
plt.barh(Q16["Frameworks"],Q16["count"])
```
