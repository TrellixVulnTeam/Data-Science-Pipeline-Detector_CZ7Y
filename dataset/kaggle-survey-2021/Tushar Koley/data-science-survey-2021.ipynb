```python
def get_avg_age(r_age):
    ages = r_age.split('-')
    if len(ages)==2:
        return (int(ages[0]) + int(ages[1]))/2
    elif r_age=='70+':
        return 70
    else:
        return r_age
    
data1=data.iloc[1:] # Removing the fisrt row is question description
def countplot(x, hue, **kwargs):
    sns.countplot(x=x, hue=hue, **kwargs)
```

```python
columns=data.columns
def plot_bar(start_index, end_index, title=None, rotation=45, figsize=(10,7)):
# start_index = 221
# end_index = 230
    data_dict = dict()
    for i in range(start_index,end_index):
        desc = ((data.iloc[0][i]).split('-')[-1]).strip()
        column = columns[i]
        desc = desc[:50]
        data_dict[desc] = data[column].count()

    sorted_dict=dict(sorted(data_dict.items(), key=lambda item: item[1], reverse=True))
    y=list(sorted_dict.values())[1:]
    l=np.sum(y)
    x=list(sorted_dict.keys())[1:]
    fig = plt.figure(figsize=figsize)
    ax=sns.barplot(x=x, y=y, palette='coolwarm')
    #plt.xlabel('Language Used')
    plt.title(title)
    plt.xticks(rotation=rotation)
    
    h_text = int(0.01*max(y))

    for p in ax.patches:
       ax.annotate('{}%'.format(np.round(p.get_height()*100/l,2)), (p.get_x()+0.1, p.get_height()+h_text))
    plt.tight_layout()

```
