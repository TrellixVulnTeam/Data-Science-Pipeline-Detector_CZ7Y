<h3> 2.1 Age Distribution

```python
plt.rcParams['figure.figsize'] = (16,9)
plt.rcParams['axes.edgecolor'] = 'black'
plt.rcParams['axes.linewidth'] = 1.5
plt.rcParams['figure.frameon'] = True
plt.rcParams['axes.spines.top'] = False
plt.rcParams['axes.spines.right'] = False
plt.rcParams["font.family"] = "monospace"
```

```python
sns.countplot(x = dataframe['Q1'].sort_values(),
              color=colors[2],
              edgecolor='white',
              linewidth=1.5,
              saturation=1.5)

plt.ylabel('Count', weight='semibold', fontname = 'Georgia')
plt.xlabel('Age Group', weight='semibold', fontname = 'Georgia')
plt.tick_params(axis='x', which='major', labelsize=10)

plt.suptitle('Age Distribution', fontname = 'Georgia', weight='bold')
plt.show()
```

<blockquote><p><strong> üìå Takeaway :</strong></p>
<ul>
<li>The data realated field is popular among 18-29.</li>
<li>The reason for the lower distribution as the age increases must be due to popularity of data jobs in recent years.
</ul> 

<h3> 2.2 Gender Distribution

```python
import plotly.express as px

pie_col = ["#CF673A","#E4C56F","#97A37B","#BABABA","#E8DCCC","#1274A1"]

fig = px.pie(values= dataframe['Q2'].value_counts().values, names= dataframe['Q2'].value_counts().index)

fig.update_traces(hoverinfo='label+percent',
                  textinfo='value',
                  textfont_size=20,
                  marker=dict(colors=pie_col,
                              line=dict(color='#000000', width=1)))
fig.show()
```

<blockquote><p><strong> üìå Takeaway :</strong></p>
<ul>
<li>There are Large number of men on data science field.</li>
<li>As expected from the previous surveys thre small number of female kagglers than male.</li>
</ul> 

<h3> 2.3 Countries Participated

```python
scale =[[0, '#E4C56F'],
             [0.5,'#455C4F'],
             [1.0, '#CF673A']]


settings = [ dict(
        type = 'choropleth',
        autocolorscale = False,
        colorscale = scale,
        showscale = True,
        locations = dataframe.Q3.value_counts().index,
        z = dataframe.Q3.value_counts().values,
        locationmode = 'country names',
        text = dataframe.Q3.value_counts().index,
        marker = dict(
            line = dict(color = '#E4C56F', width = 0.5)),
            colorbar = dict(autotick = True, tickprefix = '', 
            title = 'Participants')
            )
       ]

layout = dict(
    title = 'Number of People Participated in Survey',
    geo = dict(
        showframe = True,
        showocean = True,
        oceancolor = '#1274A1',
        projection = dict(
        type = 'orthographic',
            rotation = dict(
                    lon = 60,
                    lat = 10),
        ),
        lonaxis =  dict(
                showgrid = False,
                gridcolor = '#6e6d6d'
            ),
        lataxis = dict(
                showgrid = False,
                gridcolor = '#6e6d6d'
                )
            ),
        )
fig = dict(data=settings, layout=layout)
py.iplot(fig, validate=False, filename='worldmap')
```

<h3> 2.4 Level of Formal Education

```python
from textwrap import wrap

A = sns.countplot(x= dataframe['Q4'],
              color=colors[2],
              edgecolor='white',
              linewidth=1.5,
              saturation=1.5)

labels = [ '\n'.join(wrap(l, 12)) for l in dataframe['Q4'].unique()]

plt.ylabel('Count', weight='semibold', fontname = 'Georgia')
plt.xlabel('Education', weight='semibold', fontname = 'Georgia')

plt.suptitle('Formal Education', fontname = 'Georgia', weight='bold')

plt.tick_params(axis='x', which='major', labelsize=10)

A.set_xticklabels(labels, fontsize=10)

plt.show()
```

<h3> 2.5 Current Role

```python
import plotly.graph_objs as go

edu = dataframe['Q5'].value_counts()

trace = go.Scatter(
    x=edu.index,
    y=edu.values,
    mode='markers',
    marker=dict(
        sizemode = 'diameter',
        sizeref = 30,
        size = edu.values,
        color = edu.values,
        colorscale='Fall',
        showscale=True
    ),
)

layout = go.Layout(
    title='Current Role'
)

data = [trace]
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename="Current Role")
```

<h3> 2.6 Coding Experience

```python
import plotly.graph_objs as go

edu = dataframe['Q6'].value_counts()

trace = go.Scatter(
    x=edu.index,
    y=edu.values,
    mode='markers',
    marker=dict(
        sizemode = 'diameter',
        sizeref = 30,
        size = edu.values,
        #color = np.random.randn(500), #set color equal to a variable
        color = edu.values,
        colorscale='Fall',
        showscale=True
    ),
)

layout = go.Layout(
    title='Coding Experience'
)

data = [trace]
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename="LearningDataScience")
```
