## 5.Gender Distribution in the Top 20 countries 

```python
women_con = women.loc[:, 'Q2':'Q3']
women_con.head()
```

```python
men_con = men.loc[:, 'Q2':'Q3']
men_con.head()
```

```python
w_con = women_con.groupby('Q3').count()
w_con = w_con.reset_index()
w_con
```

```python
m_con = men_con.groupby('Q3').count()
m_con = m_con.reset_index()
m_con
```

```python
w_m_merge = pd.merge(w_con,m_con, on = "Q3")
w_m_merge.columns = ['Country','Women','Men']
w_m_merge
```

```python
# To plot for the top 20 countries
gen_con = w_m_merge
gen_con['Total'] = gen_con['Men'] + gen_con['Women']
gen_con = gen_con.sort_values(by='Total', ascending = False)
gen_con = gen_con.iloc[:20,:]
gen_con['Country'] = gen_con['Country'].replace(["United States of America","United Kingdom of Great Britain and Northern Ireland"]
                   ,["USA", "UK"])
gen_con
```

```python
x1 = gen_con['Country']
y1 = gen_con["Men"]
y2 = gen_con["Women"]
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(20,7))
plt.bar(x1, y1, label = "Men", width = 0.5, color = '#c1e7ff')
plt.bar(x1, y2, label = "Women", width = 0.5 , bottom = y1, color = '#5886a5')
plt.xlabel('Countries', fontsize = 20)
plt.ylabel('Number of Users',fontsize = 15)
plt.xticks(rotation = 90)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
for x,y in zip(x1,y1):
    plt.text(x, y-250, '%d' % y, ha='center' , va= 'bottom')
for x,y,z in zip(x1,y2,y1):
    plt.text(x, y+z, '%d' % y, ha='center' , va= 'bottom')
plt.title ('Gender distribution of users among Countries', fontsize = 20)
plt.legend(prop={"size":20})
plt.show()
```

## 8. Gender and Job

```python
m_job = men[['Q2', 'Q5']]
m_job.columns = ['Men', 'Job']
m_job = m_job.groupby('Job').count()
m_job = m_job.reset_index()
m_job
```

```python
w_job = women[['Q2', 'Q5']]
w_job.columns = ['Women', 'Job']
w_job = w_job.groupby('Job').count()
w_job = w_job.reset_index()
w_job
```

```python
all_job = pd.merge(w_job,m_job, on = "Job")
all_job.columns = ['Job','Women','Men']
all_job
```

```python
x1= all_job["Job"]
y1= all_job['Men']
y2 = all_job['Women']
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(20,10))
ax = plt.axes()
ax.set_facecolor('white')
plt.barh(x1,y1,label = "Men",color = '#ffd000')
plt.barh(x1,y2,label = "Women", left = y1 , color = '#fff1b3')
plt.xlabel('Count of Gender', fontsize = 15)
plt.ylabel('Job',fontsize = 15)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
for x,y in zip(x1,y1):
    plt.text(y, x, '%d' % y, ha='center' , va= 'bottom')
for x,y,z in zip(x1,y2,y1):
    plt.text(y+z, x, '%d' % y, ha='center' , va= 'bottom')
# plt.title ('$Bar $ $ Chart$')
plt.legend(prop={"size":20})
plt.show()
```

## 19-22 Industry & Size of Team

```python
data = kaggle[['Q20','Q22']]
fourth = (data[data['Q22']== '5-9'].groupby('Q20').count()).reset_index()
fifth = (data[data['Q22']== '10-14'].groupby('Q20').count()).reset_index()
sixth = (data[data['Q22']== '15-19'].groupby('Q20').count()).reset_index()
seventh = (data[data['Q22']== '20+'].groupby('Q20').count()).reset_index()

plt.figure(figsize=(15,4))
ax = plt.axes()
ax.set_facecolor('white')
b = plt.bar(fourth['Q20'], fourth['Q22'],color = "#599db4")
plt.xlabel('Industry')
plt.ylabel('Count of Teams of Size 5-9')
plt.title('Count of Teams of Size 5-9 Across Different Industries')
plt.xticks(rotation=90)
for bar in b:
    yval = bar.get_height()
    plt.text(bar.get_x()-0.25 + bar.get_width()/2.0, yval, int(yval), va='bottom')
plt.show()


plt.figure(figsize=(15,4))
ax = plt.axes()
ax.set_facecolor('white')
i = plt.bar(fifth['Q20'], fifth['Q22'],color = '#6ca9d0')
plt.xlabel('Industry')
plt.ylabel('Count of Teams of Size 10-14')
plt.title('Count of Teams of Size 10-14 Across Different Industries')
plt.xticks(rotation=90)
for bar in i:
    yval = bar.get_height()
    plt.text(bar.get_x()-0.25 + bar.get_width()/2.0, yval, int(yval), va='bottom')
plt.show()

plt.figure(figsize=(15,4))
ax = plt.axes()
ax.set_facecolor('white')
c = plt.bar(sixth['Q20'], sixth['Q22'],color = '#6bbfd7')
plt.xlabel('Industry')
plt.ylabel('Count of Teams of Size 15-19')
plt.title('Count of Teams of Size 15-19 Across Different Industries')
plt.xticks(rotation=90)
for bar in c :
    yval = bar.get_height()
    plt.text(bar.get_x()-0.25 + bar.get_width()/2.0, yval, int(yval), va='bottom')
plt.show()

plt.figure(figsize=(15,4))
ax = plt.axes()
ax.set_facecolor('white')
d = plt.bar(seventh['Q20'], seventh['Q22'],color = '#78d3ef')
plt.xlabel('Industry')
plt.ylabel('Count of Teams of Size 20+')
plt.title('Count of Teams of Size 20+ Across Different Industries')
for bar in d:
    yval = bar.get_height()
    plt.text(bar.get_x()-0.25 + bar.get_width()/2.0, yval, int(yval), va='bottom')
plt.xticks(rotation=90)
plt.show()
```
