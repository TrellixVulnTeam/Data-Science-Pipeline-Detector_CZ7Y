```python
import seaborn as sns
from matplotlib import pyplot as plt
```

# Age distribution

```python
D1=data['Q1'].value_counts().reset_index()
D1.drop([11],axis=0,inplace=True)
D1=D1.sort_values(by='index')
D1
```

```python
Y=(D1['Q1']/D1['Q1'].sum())*100
myexplode=(0.0,0.1,0.2,0.2,0.1,0.2,0.1,0.1,0.2,0.1,0.2)
plt.style.use("fivethirtyeight")
mylabel=D1['index']
colors={'#9b2226','#ae2012','#bb3e03','#ca6702','#ee9b00','#e9d8a6','#94d2bd',
'#0a9396','#005f73','#001219','#99d98c'}
plt.pie(Y,labels=mylabel,autopct="%1.1f%%",startangle=15,explode=myexplode,shadow=True,colors=colors)
plt.axis("equal")
plt.title("Age distribution on kaggle in 2021")
plt.legend(mylabel)
plt.gcf().set_size_inches(15,8)
plt.show()
```

# Gender of the respondents

```python
D2=data['Q2'].value_counts().reset_index()
D2.drop([5],axis=0,inplace=True)
D2=D2.sort_values(by='Q2')
D2
```
