### 1- Drop any Column has null values large than 15000

```python
dic = df.isnull().sum().to_dict()

cols = []

for x in dic:
    key = x
    value_nulls = dic[x]
    
    if value_nulls < 15000:
        cols.append(key)
    
```

# Visualizations

**Lets look at Duration column.**

```python
plt.figure(figsize=(12,5))

sns.boxplot(y=df['Duration'], color='#464E2E')

plt.show()
```

As we can see at the box plot, There are a lot of **Outliears**. we have to remove it!

```python
# Remove Outliears

q75,q25 = np.percentile(df['Duration'],[75,25])
intr_qr = q75-q25

max = q75+(1.5*intr_qr)
min = q25-(1.5*intr_qr)

# df.loc[df['Duration'] < min,x] = np.nan
# df.loc[df['Duration'] > max,x] = np.nan

duration_withou_outliears = df['Duration'][(df['Duration'] > min) & (df['Duration'] < max)]
```

```python
plt.figure(figsize=(12,5))

sns.boxplot(y=duration_withou_outliears, color='#464E2E')

plt.show()
```

Now we are removed the outliears, so we can extract right analytics such as min, max, mean, and mode.

```python
print("Avarage of Duration is :", round(duration_withou_outliears.mean(),2) )
```

```python
print("Min Duration is :", round(duration_withou_outliears.min(),2) )
```

```python
print("Max Duration is :", round(duration_withou_outliears.max(),2) )
```

```python
print("Mode of Duration is :", duration_withou_outliears.mode().values[0] )
```

```python
sns.displot(duration_withou_outliears, color='#464E2E', bins=60, alpha=.86, aspect=3, height=7)

plt.title('Distribution of Duration after removeing the outliears.', fontsize=30, color='#03045E')

plt.xticks(fontsize=20, color='black')
plt.yticks(fontsize=20,  color='black')

plt.xlabel('Duration', fontsize=25, color='black')
plt.ylabel('Count', fontsize=25, color='black')

plt.show()
```
