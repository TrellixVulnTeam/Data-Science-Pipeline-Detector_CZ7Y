> #  User distribution across Geopraphy

```python
#import plotly.plotly as plt --deprecated
import plotly.graph_objs as gob
import cufflinks as cf
from plotly.offline import download_plotlyjs,init_notebook_mode, plot, iplot

#data = plotdata.reset_index(drop=True)
init_notebook_mode(connected = True)
cf.go_offline()
cols = ['In which country do you currently reside?']
#Convert value counts to DF
user_count =pd.DataFrame(df[cols].value_counts())
df_usercount = user_count.reset_index()
df_usercount.columns = ['Country', 'User_count']

#fig = px.line(data, x ='What is your age (# years)?	What is your gender? - Selected Choice	', y = 'count')
print(df_usercount)
df_usercount.iplot(kind ='barh', x='Country', y='User_count')
#df_usercount['Country'].iplot()
df_usercount.iplot(kind= 'scatter', mode='markers')

#choromap = gob.Figure(data = [data],layout = layout)
#df_usercount.iplot(choromap)
```

> #GLobal Map of Kaggle Users

```python
# iplot global map 
fig = gob.Figure(data=gob.Choropleth(
    locations = df_usercount['Country'],
    locationmode='country names',
    z = df_usercount['User_count'],
    text = df_usercount['Country'],
    colorscale = 'spectral',
    autocolorscale=False,
    reversescale=True,
    marker_line_color='darkgray',
    marker_line_width=0.5,
    colorbar_tickprefix = '',
    colorbar_title = 'Geo wide distribution',
))

fig.update_layout(
    title_text='2021 Kaggle users',
    geo=dict(
        showframe=False,
        showcoastlines=False,
        projection_type='equirectangular'
    ),
    annotations = [dict(
        x=0.55,
        y=0.1,
        xref='paper',
        yref='paper',
        text='Source:https://www.kaggle.com/c/kaggle-survey-2021',
        showarrow = False
    )]
)
iplot(fig)
```
