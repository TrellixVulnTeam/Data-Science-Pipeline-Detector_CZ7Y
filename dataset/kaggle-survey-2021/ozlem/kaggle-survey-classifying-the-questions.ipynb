It seems that column names depict the questions and some properties such as duration, age, country, etc; also, row 0 does not belong to the answers but presents the desciptions of the properties titled in the column names. Let's explore the column names and this row 0:     
        
We have 369 columns, however, not all columns belong to a single question, there are a lot of questions that spread to multiple columns since they are multiple choice type. 

There are also some other considerations as stated in "kaggle_survey_2021_methodology.pdf" file:


    
* Respondents with the most experience were asked the most questions. For example, students and unemployed persons were not asked questions about their employer. Likewise, respondents that do not write code were not asked questions about writing code.
* Follow-up questions were only asked to respondents that answered the setup question affirmatively.
  * Question 18 (which specific ML methods) was only asked to respondents that selected the relevant answer choices for Question 17 (which categories of algorithms).
  * Question 19 (which specific ML methods) was only asked to respondents that selected the relevant answer choices for Question 17 (which categories of algorithms).
  * Question 28 (which specific product) was only asked to respondents that selected more than one choice for Question 27-A (which of the following products).
  * Question 29-A (which specific AWS/Azure/GCP products) was only asked to respondents that selected the relevant answer choices for Question 27-A (which of the following companies).
  * Question 30-A (which specific AWS/Azure/GCP products) was only asked to respondents that selected the relevant answer choices for Question 27-A (which of the following companies).
  * Question 33 (which specific product) was only asked to respondents that selected more than one choice for Question 32-A (which of the following products).
  * Question 35 (which specific product) was only asked to respondents that selected more than one choice for Question 34-A (which of the following products).
  * Question 37-A (which specific product) was only asked to respondents that answered affirmatively to Question 36-A (which of the following categories of products).
* For questions about cloud computing products, students and respondents that have never spent money in the cloud were given an alternate set of questions that asked them “what products they would like to become familiar with” instead of asking them “which products they use most often”. For questions with alternative phrasing, the questions were kept separate, and question types were labeled with either an “A” or a “B” (e.g. Q29A, Q29B, ... , Q37A, Q37B).

```python
q = {}

for col in df_responses.columns.drop("Time from Start to Finish (seconds)"):
    #print(col, df_responses[col][0])
    if "Part" in col:
        colname = col[1:].split("Part")[0].replace("_","") + ":" + df_responses[col][0] 
        colname = colname.split("(Select all that apply)")[0]
        colname = colname.split("- Selected Choice -")[0] + "MULTIPLECOLUMN"
    elif "OTHER" in col:
        colname = col[1:].split("OTHER")[0].replace("_","") + ":" + df_responses[col][0]
        colname = colname.split("(Select all that apply)")[0]
        colname = colname.split("- Selected Choice -")[0] + "MULTIPLECOLUMN"       
    else:
        colname = col[1:] + ":" + df_responses[col][0]

       
        
    parts = colname.split(":")
    

        
    if "A" in parts[0]:
        key = int(parts[0].replace('A','') + "1")
        value = parts[1] + ", ALTERNATIVE PHRASING"
    elif "B" in parts[0]:
        key = int(parts[0].replace('B','') + "2")
        value = parts[1] + ", ALTERNATIVE PHRASING"
    else:
        key = int(parts[0])
        value = parts[1]
      
    
    
    if key in [18,19,28,291,292,301,302,33,35,371,372]:
        q[key]= value + ", FOllOW-UP"
    else:
        q[key]= value


questions = q.items()

questions = sorted(questions)


print("Questions:")
for q in questions:
    print(q[0],' : ', q[1])
```

                Finally, let's see which questions are single column, multiple column (multiple selection), alternative phrasing, follow-up, for professionals or for non-professionals:
        </font></div>

```python
def print_question_set(s):
    
    for key in s:
        if len(str(key))<3:
            print(key, ' : ', s[key].split("?")[0]+"?")    
        else:
            if str(key)[2] == "1":
                print(str(key)[:2]+"A", ' : ', s[key].split("?")[0]+"?")
            else:
                print(str(key)[:2]+"B", ' : ', s[key].split("?")[0]+"?")
            
            
questions_single_cols = {}
questions_multiple_cols = {}
questions_alternative_phrasing = {}
questions_follow_up = {}
questions_for_only_professionals = {}
questions_for_only_nonprofessionals = {}



for q in questions:
    if "MULTIPLECOLUMN" in q[1]:
        questions_multiple_cols[q[0]] = q[1]
    else:
        questions_single_cols[q[0]] = q[1]
    

print ("Single column questions")
print_question_set(questions_single_cols)
    
print("\n\nMultiple column questions ('Multiple selection questions)")
print_question_set(questions_multiple_cols)
    
    
    
for q in questions:
    if "ALTERNATIVE PHRASING" in q[1]:
        questions_alternative_phrasing[q[0]] = q[1]

print ("\n\nAlternatively phrased questions")
print_question_set(questions_alternative_phrasing)      
        
    
    
for q in questions:
    if "FOllOW-UP" in q[1]:
        questions_follow_up[q[0]] = q[1]

print ("\n\nFollow-up questions")
print_question_set(questions_follow_up)



for q in questions:
    if len(str(q[0]))>2:
        if str(q[0])[2]=="1":
            questions_for_only_professionals[q[0]] = q[1]
        elif str(q[0])[2]=="2":
            questions_for_only_nonprofessionals[q[0]] = q[1]

print ("\n\nQuestions for only professionals")
print_question_set(questions_for_only_professionals)

print ("\n\nQuestions for only non-professionals (students, unemployed, and respondents that have never spent any money in the cloud)")
print_question_set(questions_for_only_nonprofessionals)
                
```
