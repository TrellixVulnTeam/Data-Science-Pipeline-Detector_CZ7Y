## Percentage of programming experience among data scientsits
### Shown almost half the population have less than 3 years of programming

```python
newdf['Yrs_Programming'].value_counts(normalize=True).mul(100).round()
```

```python
waffle_counts = newdf['Yrs_Programming'].value_counts(normalize=True).mul(100).round()

prev_count = 0
# for each category,
for cat in range(waffle_counts.shape[0]):
    # get the block indices
    blocks = np.arange(prev_count, prev_count + waffle_counts[cat])
    # and put a block at each index's location
    x = blocks % 10 # use mod operation to get ones digit
    y = blocks // 10 # use floor division to get tens digit
    plt.bar(x = x, height = 0.8, width = 0.8, bottom = y)
    prev_count += waffle_counts[cat]

# aesthetic wrangling
plt.legend(waffle_counts.index, bbox_to_anchor = (1, 0.5), loc = 6)
plt.title('DS Years of Programming')
plt.axis('off')
plt.axis('square')
```

## Programming lanugage of Data Scientist compared to Age groups for different Genders
