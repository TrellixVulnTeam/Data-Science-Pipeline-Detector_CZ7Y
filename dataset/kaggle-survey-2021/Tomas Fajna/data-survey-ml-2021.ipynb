```python
world_country = survey.pivot_table(columns=['Q3'], aggfunc='size')
world_sorted = world_country.sort_values(axis=0, ascending=True, inplace=False, kind='quicksort', na_position='last', ignore_index=False, key=None)
```

```python
world_sorted.tail(10)
```

# In which country do you currently reside?

```python
world_sorted.tail(10).plot(kind='pie', ylabel='Country')
```
