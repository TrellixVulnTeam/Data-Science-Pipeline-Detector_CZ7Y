```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('../input/kaggle-survey-2021/kaggle_survey_2021_responses.csv')
```

```python
data.head(3)
```

```python
data.drop([0],inplace = True)
data = data.rename(columns={'Time from Start to Finish (seconds)': 'Duration', 'Q1': 'Age' , 'Q2':'Gender','Q3':'Country','Q4':'Education','Q5':'Job Title','Q6':'Experience Years','Q7_Part_1':'Python','Q7_Part_2':'R','Q7_Part_3':'SQL','Q7_Part_4':'C','Q7_Part_5':'C++','Q7_Part_6':'Java','Q7_Part_7':'Javascript','Q7_Part_8':'Julia','Q7_Part_9':'Swift','Q7_Part_10':'Bash','Q7_Part_11':'MATLAB','Q7_Part_12':'None','Q7_OTHER':'Other','Q8':'Recommended'})
```

```python
data['Country'].unique() 
data['Country']= data['Country'].replace(["United States of America","United Kingdom of Great Britain and Northern Ireland","I do not wish to disclose my location","United Arab Emirates"],["USA", "UK","undisclosed","UAE"])
```

```python
data['Education'].unique() 
data['Education'] =data['Education'].replace(["Some college/university study without earning a bachelorâ€™s degree"],["Some college"])
```

```python
sub_data = data.iloc[:,0:21]
sub_data
```

# Graph 1
## What is the most feequent age for whome taking the survey 

```python
unique_Age=sub_data['Age'].value_counts()#---->get unique age in column
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(7,5),edgecolor='#FFFFFF')
unique_Age.plot(kind='bar')
plt.title("Age",fontweight='bold')
```

# Graph 2
## What is the most feequent genderfor whome taking the survey 

```python
def gender_compiner (x):
    if (x == 'Prefer not to say') or (x == 'Nonbinary') or x == 'Prefer to self-describe':
        return 'Other'
    else:
        return x
    # lets test it first
sub_data['Gender'] = sub_data['Gender'].apply(gender_compiner)

gender_count = sub_data['Gender'].value_counts()

explod = (0.1,0.1, 0.1)
plt.figure(figsize = (8,8))

plt.pie(gender_count,explode=explod,labels=gender_count.index,autopct='%1.1f%%',shadow=True)
plt.title('Gender Distribution',fontweight='bold')
```

# Graph 19

```python
Q9_columns = [column for column in data.columns if 'Q9' in column]
Q9_dict = {}

for column in Q9_columns:
    Q9_dict.update(dict(data[column].value_counts()))

Q9_df = pd.DataFrame(Q9_dict.items(), columns = ['IDE', 'counts'])

Q9_df["counts"][0] = Q9_df["counts"][0] + Q9_df["counts"][10]
Q9_df.drop(index=10, axis=1, inplace=True)

labels = list(Q9_df["IDE"])
values = list(Q9_df["counts"])
fig = go.Figure(data=[go.Pie(labels=labels, values=values, hole=.3, textinfo='label+percent',
                             insidetextorientation='radial'
                            )])
fig.show()
```
