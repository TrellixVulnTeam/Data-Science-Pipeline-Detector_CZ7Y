# Level of education of participants

```python
education_level=df['Q4'].loc[1:]
```

```python
education_prop = education_level.value_counts(normalize =True)
education_prop
```

```python
plt.figure(figsize = (9,9))
plt.style.use('ggplot')
education_prop.plot(kind = 'barh')
pos,label = plt.yticks()

for x,y in zip(pos,label):
        count = education_prop[y.get_text()]
        stringvalue = '{:0.02f}'.format(count)
        plt.text(count+0.01, x, stringvalue, ha='center' , va= 'bottom')
plt.xlabel('levels of education')
plt.ylabel('Proportion  of educational level')
plt.title('Distribution of levels of education')
```

We can see that more than half of the people intereeted in data science taht participate in the survey has strong academic background(matster and bachelors degree) (in order to generalize to whole population we need statistical test)
