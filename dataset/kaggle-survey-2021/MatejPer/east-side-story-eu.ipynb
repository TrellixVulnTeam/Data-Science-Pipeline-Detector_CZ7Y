# Introduction

Kaggle 2021 survey gives us an another opportunity to investigate trends in data world. In this notebook we will focus on the kagglers from the eastern european countries and we will make comparison with the kagglers from India, USA, rest of Europe and rest of the world. 
Countries that we have included as eastern european are Russia, Turkey, Poland, Ukraine, Greece, Czech Republic, Romania and Belarus. Some of them have gone through similar processes of getting independence and economical transition in the late years of the 20th century. Opening to global markets and access to world wide web have been an opportunity for them to catch up with the west.

## Part I

In the first part of the notebook we will look into the structure of kaggle users in east Europe. 
* The **highest** number of kagglers are coming from **Russia** and the **lowest** number from **Belarus**.
* **19% of kagglers** are in the age group **between 25 and 29 years**, making it the most represented age group
* There are **over 82% men** and **almost 16% of women**
* **85,3% are highly educated**, having at least bachelor's degree
* The most represented **job role** is the one of **data scientist**, followed with students

*  Comparison of distribution of experience is showing an interesting trend - **eastern europeans are more similar to kagglers from India** then to kagglers from rest of Europe
