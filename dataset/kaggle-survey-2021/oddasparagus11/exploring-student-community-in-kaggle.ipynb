```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from matplotlib.pyplot import figure

df = pd.read_csv("/kaggle/input/kaggle-survey-2021/kaggle_survey_2021_responses.csv", low_memory = False)
df.drop(index=df.index[0], axis=0,inplace=True)
```

Hi, everyone!! If you are an Indian student and have no other job to do like me(which is highly likely given that more than 65% of the students who have taken the survey are from India), this is a analysis of what the student community in kaggle is like and what could make you stand out and about whether we are all the same or we have our differences with respect to other factors like country, degree we are pursuing etc.. 

```python
Q5_distrib = pd.DataFrame(df['Q5'].value_counts())
Q5_distrib.reset_index(inplace=True)
Q5_distrib_grey = Q5_distrib.copy() 
sns.set(style="whitegrid")
f,ax = plt.subplots(figsize=(10, 8))
colors = ['red','b','b','b','b','b','b','b','b','b','b','b','b','b','b']
sns.barplot(x="Q5", y="index", data=Q5_distrib, palette = colors)
sns.despine(left=True, bottom=True)
sns.set(style="whitegrid")
sns.barplot(x="Q5", y="index", data=Q5_distrib_grey,
            palette =colors)
sns.despine(left=True, bottom=True)
_ = plt.text(6850, 0.1, '6804', fontsize=12, color='green')
_ = ax.set_title('Distribution Based on Designation', fontsize=20)
ax.xaxis.grid(False)
ax.yaxis.grid(False)
ax.set_xticks([])
ax.xaxis.set_label_text("")
ax.yaxis.set_label_text("")
_ = plt.setp(f.patches, linewidth=0.5) 
```

Since, I was only interested to see the differences and similarities in the student community, I only considered those data points in the dataset and removed others. Eventhough most of the students in kaggle are male(around 72%), as you can see from the graph below that the degrees pursuied by both of them doesnt seem to change much. Around 52% is doing their bachelors and 25% in their Master's degree right now.

```python
from plotly.offline import init_notebook_mode, iplot 
import plotly.graph_objs as go
import plotly.offline as py
import pycountry
male_2019 = df[df['Q2']=='Man']
female_2019 = df[df['Q2']=='Woman']

topn = 10
count_male = male_2019['Q4'].value_counts()[:topn].reset_index()
count_female = female_2019['Q4'].value_counts()[:topn].reset_index()

pie_men = go.Pie(labels=count_male['index'],values=count_male['Q4'],name="Men",hole=0.4,domain={'x': [0,0.46]})
pie_women = go.Pie(labels=count_female['index'],values=count_female['Q4'],name="Women",hole=0.5,domain={'x': [0.52,1]})

layout = dict(title = 'Degrees Pursuied by Students in Kaggle', font=dict(size=12), legend=dict(orientation="h"),
              annotations = [dict(x=0.2, y=0.5, text='Men', showarrow=False, font=dict(size=20)),
                             dict(x=0.8, y=0.5, text='Women', showarrow=False, font=dict(size=20)) ])

fig = dict(data=[pie_men, pie_women], layout=layout)
py.iplot(fig)
```
