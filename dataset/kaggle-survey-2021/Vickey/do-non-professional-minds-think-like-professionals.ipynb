```python
# function to generate dataframe for plotly sunburst

def generate_dataframe_sunburst(df,col_name,experience):
  col_values=[]
  for col, values in df.iteritems():
    col_values+=[col]*(values.notnull().sum())
  data_column = pd.Series(col_values)
  experience_column = pd.Series([experience]*data_column.shape[0])

  df_sunburst = pd.DataFrame({col_name:data_column,
                                    "Experience":experience_column})
  return df_sunburst
```

```python
#generating dataframes from different questions for plotly sunburst

cloud_platform_pro_df = generate_dataframe_sunburst(d_27A,'Cloud Platform','Professional')
cloud_platform_non_pro_df = generate_dataframe_sunburst(d_27B,'Cloud Platform','Non-Professional')
cloud_platform_sunburst_df = pd.concat([cloud_platform_pro_df,cloud_platform_non_pro_df],ignore_index=True)

cloud_computing_pro_df = generate_dataframe_sunburst(d_29A,'Cloud Computing','Professional')
cloud_computing_non_pro_df = generate_dataframe_sunburst(d_29B,'Cloud Computing','Non-Professional')
cloud_computing_sunburst_df = pd.concat([cloud_computing_pro_df,cloud_computing_non_pro_df],ignore_index=True)

ml_products_pro_df = generate_dataframe_sunburst(d_31A,'ML Products','Professional')
ml_products_non_pro_df = generate_dataframe_sunburst(d_31B,'ML Products','Non-Professional')
ml_products_sunburst_df = pd.concat([ml_products_pro_df,ml_products_non_pro_df],ignore_index=True)

bigdata_pro_df = generate_dataframe_sunburst(d_32A,'BigData Products','Professional')
bigdata_non_pro_df = generate_dataframe_sunburst(d_32B,'BigData Products','Non-Professional')
bigdata_sunburst_df = pd.concat([bigdata_pro_df,bigdata_non_pro_df],ignore_index=True)

bi_tools_pro_df = generate_dataframe_sunburst(d_34A,'BI tools','Professional')
bi_tools_non_pro_df = generate_dataframe_sunburst(d_34B,'BI tools','Non-Professional')
bi_tools_sunburst_df = pd.concat([bi_tools_pro_df,bi_tools_non_pro_df],ignore_index=True)

auto_ml_pro_df = generate_dataframe_sunburst(d_36A,'Auto ML','Professional')
auto_ml_non_pro_df = generate_dataframe_sunburst(d_36B,'Auto ML','Non-Professional')
auto_ml_sunburst_df = pd.concat([auto_ml_pro_df,auto_ml_non_pro_df],ignore_index=True)

auto_ml_tools_pro_df = generate_dataframe_sunburst(d_37A,'Auto ML tools','Professional')
auto_ml_tools_non_pro_df = generate_dataframe_sunburst(d_37B,'Auto ML tools','Non-Professional')
auto_ml_tools_sunburst_df = pd.concat([auto_ml_tools_pro_df,auto_ml_tools_non_pro_df],ignore_index=True)

manage_ml_pro_df = generate_dataframe_sunburst(d_38A,'Manage ML','Professional')
manage_ml_non_pro_df = generate_dataframe_sunburst(d_38B,'Manage ML','Non-Professional')
manage_ml_sunburst_df = pd.concat([manage_ml_pro_df,manage_ml_non_pro_df],ignore_index=True)
```

```python
#plotting sunburst for cloud platforms

fig = px.sunburst(cloud_platform_sunburst_df, path=['Experience','Cloud Platform'],template='ggplot2',height=600,
                  hover_name='Cloud Platform',title='Popular Cloud Platforms')
fig.update_traces(textinfo='label+percent entry')
fig.show()
```

* Number of non-professionals wanting to use Amazon Web Services (AWS), Google Cloud Platform (GCP), Microsoft Azure is approximately similar to number of professionals.
* There is a slight increase in number of non-professionals wanting to use IBM, Oracle, SAP compared to professionals.
