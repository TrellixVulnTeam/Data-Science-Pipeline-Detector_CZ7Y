**<font color = CadetBlue size = 4 >- Background -</font>**

**In the first section, I will discuss the background of kagglers, including age, gender, country and education.**

```python
# Age
df1 = dfc['Q1'].value_counts().sort_values(ascending = True).reset_index()
df1 = df1.rename(columns={'index':'Age', 'Q1':'Count'})
df1['percent'] = (df1['Count']*100 / df1['Count'].sum()).round(1).astype(str) + '%'

# Gender
df2 = dfc['Q2'].value_counts().sort_values(ascending = True).reset_index()
df2 = df2.rename(columns={'index':'Gender', 'Q2':'Count'})
df2['percent'] = (df2['Count']*100 / df2['Count'].sum()).round(1).astype(str) + '%'

# Country
df3 = dfc['Q3'].value_counts().sort_values(ascending = True).reset_index()
df3 = df3.rename(columns={'index':'Country', 'Q3':'Count'})
df3['percent'] = (df3['Count']*100 / df3['Count'].sum()).round(1).astype(str) + '%'

# Education
df4 = dfc['Q4'].value_counts().sort_values(ascending = True).reset_index()
df4 = df4.rename(columns={'index':'Education', 'Q4':'Count'})
df4['percent'] = (df4['Count']*100 / df4['Count'].sum()).round(1).astype(str) + '%'
df4['Education'] = df4['Education'].replace("Some college/university study without earning a bachelor‚Äôs degree","Some college without a degree")

fig = go.Figure()

fig.add_trace(go.Bar(x = df1['Count'],
                     y = df1['Age'],
                     orientation='h',
                     name = 'Age',
                     text= df1['percent'],
                     texttemplate='<b>%{text}<b>', 
                     hovertemplate = 'Age: <b>%{y}</b><br>Count: <b>%{x}</b><br>',
                     textposition='outside',
                     textfont_size=11,
                     cliponaxis = False,
                     marker = dict(color = 'LightCoral')))

fig.add_trace(go.Bar(x = df2['Count'],
                     y = df2['Gender'],
                     orientation='h',
                     name = 'Gender',
                     text= df2['percent'],
                     texttemplate='<b>%{text}<b>', 
                     hovertemplate = 'Gender: <b>%{y}</b><br>Count: <b>%{x}</b><br>',
                     textposition='outside',
                     textfont_size=11,
                     cliponaxis = False,
                     marker = dict(color = 'MediumPurple')))

fig.add_trace(go.Bar(x = df3['Count'],
                     y = df3['Country'],
                     orientation='h',
                     name = 'Country',
                     text= df3['percent'],
                     texttemplate='<b>%{text}<b>', 
                     hovertemplate = 'Country: <b>%{y}</b><br>Count: <b>%{x}</b><br>',
                     textposition='outside',
                     textfont_size=11,
                     cliponaxis = False,
                     marker = dict(color = 'CadetBlue')))

fig.add_trace(go.Bar(x = df4['Count'],
                     y = df4['Education'],
                     orientation='h',
                     name = 'Education',
                     text= df4['percent'],
                     texttemplate='<b>%{text}<b>', 
                     hovertemplate = 'Education: <b>%{y}</b><br>Count: <b>%{x}</b><br>',
                     textposition='outside',
                     textfont_size=11,
                     cliponaxis = False,
                     marker = dict(color = 'SteelBlue')))

fig.update_layout( 
    updatemenus=[ 
        dict( 
            direction="down", 
            pad={"r": 10, "t": 10},
            showactive=True, 
            x=0.155, 
            xanchor="left",
            y=1.09, 
            yanchor="top",
            font = dict(color = 'Indigo',size = 14),
            buttons=list([ 
                dict(label="All", 
                     method="update", 
                     args=[ {"visible": [True,True,True,True]}, 
                            {'showlegend' : True} 
                        ]),
                dict(label="Age", 
                     method="update", 
                     args=[ {"visible": [True,False,False,False]}, 
                            {'showlegend' : False} 
                        ]), 
                dict(label="Gender", 
                     method="update", 
                     args=[ {"visible": [False,True,False,False]}, 
                            {'showlegend' : False} 
                        ]),
                dict(label='Country', 
                     method="update", 
                     args=[ {"visible": [False,False,True,False]}, 
                            {'showlegend' : False} 
                     ]), 
                dict(label='Education', 
                     method="update", 
                     args=[ {"visible": [False, False,False,True]}, 
                            {'showlegend' : False}]),
            ]),
        )])

fig.update_layout(
    annotations=[
        dict(text="Choose:", showarrow=False,
             x=0.1, y=1.065, yref="paper", align="right",
             font=dict(size=16,color = 'DarkSlateBlue'))])

fig.update_layout(title ="The Distribution of <b>Ages,Gender,Countries and Educations<b>",
                  title_x = 0.5,
                  title_font = dict(size = 20, color = 'MidnightBlue'),
                  height = 700)

fig.show()
```

**From the above charts, we can see that :**
* **More than 80% of kagglers are man. However, there are few kagglers are woman in these countries.**
* **Most of kagglers are between 22 to 29 years of age, but the proportion of kagglers aged 55 years or older is under 6%.**
* **Most of kagglers live in Japan. It is interested that the population of China is much more than Japan, but the number of kagglers in Japan is more than in China. It is worth exploring the reasons further if it is possible.**
* **Most of kagglers possess a Master's degree. As we can see that over 70% of kagglers possess a Master's degree or a Bachelor‚Äôs degree.**

```python
# Machine learning experience(Year) 
df15 = dfc['Q15'].value_counts().sort_values(ascending = True).reset_index()
df15 = df15.rename(columns={'index':'ML Experience', 'Q15':'Count'})
df15['percent'] = (df15['Count']*100 / df15['Count'].sum()).round(1).astype(str) + '%'
df15['ML Experience'] = df15['ML Experience'].replace("I do not use machine learning methods","I do not use ML methods")

# Machine learning framework
df16= dfc[[col for col in dfc if col.startswith('Q16')]]
df16.columns = ["Scikit-learn","TensorFlow","Keras","PyTorch",
                "Fast.ai", "MXNet", "Xgboost", "LightGBM", "CatBoost", "Prophet",
                "H2O 3", "Caret", "Tidymodels","JAX","PyTorch Lightning","Huggingface",
                "None","Other"]

df16 = df16.notnull().sum().to_frame().reset_index()
df16 = df16.rename(columns = {"index":"ML Framework",0:"Count"})
df16 = df16.sort_values(by = ['Count'],ascending = True)
df16['percent'] = (df16['Count']*100 / df16['Count'].sum()).round(1).astype(str) + '%'

# Machine learning algorithm
df17= dfc[[col for col in dfc if col.startswith('Q17')]]
df17.columns = ["Linear/Logistic Regression", "Decision Trees/Random Forests", 
                "Gradient Boosting Machines", 
                "Bayesian Approaches", "Evolutionary Approaches", 
                "Dense Neural Networks","Convolutional Neural Networks", 
                "Generative Adversarial Networks","Recurrent Neural Networks", 
                "Transformer Networks", "None", "Other"]
df17 = df17.notnull().sum().to_frame().reset_index()
df17 = df17.rename(columns = {"index":"ML Algorithm",0:"Count"})
df17 = df17.sort_values(by = ['Count'],ascending = True)
df17['percent'] = (df17['Count']*100 / df17['Count'].sum()).round(1).astype(str) + '%'

    
fig = go.Figure()

fig.add_trace(go.Bar(x = df15['Count'],
                     y = df15['ML Experience'],
                     orientation='h',
                     name = 'ML Experience',
                     text= df15['percent'],
                     texttemplate='<b>%{text}<b>', 
                     hovertemplate = 'ML Experience: <b>%{y}</b><br>Count: <b>%{x}</b><br>',
                     textposition='outside',
                     textfont_size=11,
                     cliponaxis = False,
                     marker = dict(color = 'CadetBlue')))

fig.add_trace(go.Bar(x = df16['Count'],
                     y = df16['ML Framework'],
                     orientation='h',
                     name = 'ML Framework',
                     text= df16['percent'],
                     texttemplate='<b>%{text}<b>', 
                     hovertemplate = 'ML Framework: <b>%{y}</b><br>Count: <b>%{x}</b><br>',
                     textposition='outside',
                     textfont_size=11,
                     cliponaxis = False,
                     marker = dict(color = 'LightCoral')))

fig.add_trace(go.Bar(x = df17['Count'],
                     y = df17["ML Algorithm"],
                     orientation='h',
                     name = "ML Algorithm",
                     text= df17['percent'],
                     texttemplate='<b>%{text}<b>', 
                     hovertemplate = "ML Algorithm: <b>%{y}</b><br>Count: <b>%{x}</b><br>",
                     textposition='outside',
                     textfont_size=11,
                     cliponaxis = False,
                     marker = dict(color = 'SteelBlue')))

                                   
fig.update_layout( 
    updatemenus=[ 
        dict( 
            direction="down", 
            pad={"r": 10, "t": 10},
            showactive=True, 
            x=0.16, 
            xanchor="left",
            y=1.09, 
            yanchor="top",
            font = dict(color = 'Indigo',size = 14),
            buttons=list([ 
                dict(label="All", 
                     method="update", 
                     args=[ {"visible": [True,True,True]}, 
                            {'showlegend' : True} 
                        ]),
                dict(label="ML Experience", 
                     method="update", 
                     args=[ {"visible": [True,False,False]}, 
                            {'showlegend' : False} 
                        ]),
                dict(label="ML Framework", 
                     method="update", 
                     args=[ {"visible": [False,True,False]}, 
                            {'showlegend' : False} 
                        ]), 
                dict(label="ML Algorithm", 
                     method="update", 
                     args=[ {"visible": [False,False,True]}, 
                            {'showlegend' : False} 
                     ]),  
            ]),
        )])

fig.update_layout(
    annotations=[
        dict(text="Choose:", showarrow=False,
             x=0.1, y=1.065, yref="paper", align="right",
             font=dict(size=16,color = 'DarkSlateBlue'))])

fig.update_layout(title ="The Distribution of <b>ML Experience,ML Frameworks and ML Algorithms<b>",
                  title_x = 0.5,
                  title_font = dict(size = 16, color = 'MidnightBlue'),
                  height = 700)

fig.show()
```
