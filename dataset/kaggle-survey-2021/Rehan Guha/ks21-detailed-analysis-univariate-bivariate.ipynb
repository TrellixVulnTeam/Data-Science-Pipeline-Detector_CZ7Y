# Univaraite Analysis

## Q0: How much time was spend on the survery

 - If the time is too less we can assume that the form was filled in hurry and the responses might be random.
 - If the duration is too high we can we say someone was working on other things and filling the form.

```python
df['Time from Start to Finish (seconds)'] = pd.to_numeric(df['Time from Start to Finish (seconds)'])
```

```python
df['Time from Start to Finish (seconds)']
```

```python
df['Time from Start to Finish (seconds)'].plot(kind='hist', logy='sym')
```

```python
df['Time from Start to Finish (seconds)'].plot(kind='kde', logx='sym', rot=45)
```
