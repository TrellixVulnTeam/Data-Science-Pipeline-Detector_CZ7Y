```python
def con(st1):
    di = {}
    for i in range(1,12):
    #print(data['Q7_Part_1'].value_counts().sort_values(ascending=False)) 
    #data['Q7_Part_1'].value_counts().sort_values(ascending=False).plot(kind='bar')
        st = f'{st1}{i}'
        temp = data[st]
        temp = temp.dropna()
        key = temp.value_counts().sort_values(ascending=False)[0]
        val = np.unique(temp)[0]
        #print(key)
        #print(val)
        di[val] = key
    return di
#print(di)
```

```python
di = con('Q7_Part_')
di = dict(sorted(di.items(),key=lambda item:item[1], reverse=True))
names = list(di.keys())
values = list(di.values())
plt.bar(range(len(di)), values, tick_label=names)
plt.xticks(rotation=90)
plt.show()
```
