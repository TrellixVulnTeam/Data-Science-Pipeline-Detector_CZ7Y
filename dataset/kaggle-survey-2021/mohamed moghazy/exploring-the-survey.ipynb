```python
def country_cleaning(x):
    if x == 'United States of America':
        x = 'USA'
    elif x == 'United Kingdom of Great Britain and Northern Ireland':
        x = 'UK'
    elif x == 'Iran, Islamic Republic of...':
        x = 'Iran'
    elif x == 'Hong Kong (S.A.R.)':
        x = 'Hong Kong'
    elif x == 'I do not wish to disclose my location':
        x = 'other'
    elif x == 'Viet Nam':
        x = 'Vietnam'
    elif x == 'United Arab Emirates':
        x = 'UA'
    return x


df['Q3'] = df['Q3'].map(lambda x : country_cleaning(x))
df['Q3'].value_counts()
```

```python
def data_science_salary(df):
    salary={}
    for name in df['Q3'].value_counts().index:
        salary[name]={}
        df1=df[df['Q3']==name]
        salary[name]['Man']=df1[df1['Q2']=='Man']['Q25'].value_counts()
        salary[name]['Woman']=df1[df1['Q2']=='Woman']['Q25'].value_counts()
        salary_df=pd.DataFrame(salary[name]).fillna(0)
        salary_df['sum']=salary_df.sum(axis=1)
        salary_df=salary_df.T/salary_df['sum']
        salary_df=salary_df[:-1]
        salary_df=salary_df.T
        salary[name]=salary_df
    return salary
        
```

```python
ide={}
for i in range(12):
    value=df['Q9_Part_'+str(i+1)].value_counts().values[0]
    text=df['Q9_Part_'+str(i+1)].value_counts().index[0]
    ide[text]=value
```

```python
def general(question_number,df,df_man,df_woman):
    gp,gp_man,gp_woman={},{},{}
    
    for i in range(12):
        value=df[question_number+'_Part_'+str(i+1)].value_counts().values[0]
        text=df[question_number+'_Part_'+str(i+1)].value_counts().index[0]
        gp[text]=value
        #############################
        value_man=df_man[question_number+'_Part_'+str(i+1)].value_counts().values[0]
        text_man=df_man[question_number+'_Part_'+str(i+1)].value_counts().index[0]
        gp_man[text_man]=value_man
        ##########################
        value_woman=df_woman[question_number+'_Part_'+str(i+1)].value_counts().values[0]
        text_woman=df_woman[question_number+'_Part_'+str(i+1)].value_counts().index[0]
        gp_woman[text_woman]=value_woman
    return gp,gp_man,gp_woman
```

```python
def adjust_dataframe(dictionary,question_name,name):
    return pd.DataFrame(dictionary.items(),columns=[question_name,name])

df10=adjust_dataframe(general('Q9',df,df_man,df_woman)[0],'IDE','Population')
df11=adjust_dataframe(general('Q9',df,df_man,df_woman)[1],'IDE','Man')
df12=adjust_dataframe(general('Q9',df,df_man,df_woman)[2],'IDE','Woman')
```

```python
df12
```

```python
def merger(first,second,third,common):
    return pd.merge(first,second,on=common).merge(third,on=common)
df_ide=merger(df10,df11,df12,'IDE')
df_ide.index=df_ide['IDE']
df_ide.drop('IDE',axis=1,inplace=True)
```

```python
def horizontal_bar_chart(ax,data,col_name,color,title):
    ax.barh(data.index,data[col_name].values,color=color)
    
    for index in data[col_name].index:
        ax.annotate(data[col_name].loc[index],xy=(data[col_name].values.sum(),index),fontsize=18,color="black",fontweight='bold')
        
    ax.set_xticks([])
    ax.set_yticks(data.index)
    ax.set_yticklabels(data.index,fontsize=15)
    ax.legend()
    ax.set_title(title,fontsize=20)
    plt.show()
```
