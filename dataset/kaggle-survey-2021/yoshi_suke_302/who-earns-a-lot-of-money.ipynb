```python
def get_mean_salary(string):
    if string is np.nan:
        return None
    stripped = string.replace('$', '').replace(',', '').replace('>', '')
    splitted = stripped.split('-')
    splitted_int = [int(s) for s in splitted]
    mean_salary = sum(splitted_int) / len(splitted_int)
    return mean_salary
```

```python
df_new.loc[:, 'Q25_mean'] = df['Q25'].map(get_mean_salary)
```

```python
# Q1
def get_mean_age(string):
    stripped = string.replace('+', '')
    splitted = stripped.split('-')
    splitted_int = [int(s) for s in splitted]
    mean_age = sum(splitted_int) / len(splitted_int)
    return mean_age
```

```python
# Q2
def get_sex_bool(string):
    if string == 'Man':
        return 1
    elif string == 'Woman':
        return 0
    else:
        return np.nan
```

```python
# Q4
education_level_dict = {
    'No formal education past high school':0,
    'Some college/university study without earning a bachelor’s degree':1,
    'Bachelor’s degree':2,
    'Master’s degree':3,
    'Doctoral degree':4,
    'Professional doctorate':5,
    'I prefer not to answer':np.nan
}
def get_education_level_int(string):
    if string is np.nan:
        return np.nan
    else:
        return education_level_dict[string]
```

```python
# Q6
def get_mean_experience(string):
    if string is np.nan:
        return np.nan
    elif string == 'I have never written code':
        return 0
    else:
        stripped = string.replace('+', '').replace('years', '').replace('<', '').replace(' ', '')
        splitted = stripped.split('-')
        splitted_int = [int(s) for s in splitted]
        mean_experience = sum(splitted_int) / len(splitted_int)
        return mean_experience
```

```python
# Q13
tpu_experience_dict = {
    'Never':0,
    'Once':1,
    '2-5 times':3.5,
    '6-25 times':15.5,
    'More than 25 times':25,
}

def get_tpu_experience_int(string):
    if string is np.nan:
        return np.nan
    else:
        return tpu_experience_dict[string]
```

```python
# Q15
ml_experience_dict = {
    'I do not use machine learning methods':0,
    'Under 1 year':1,
    '1-2 years':1.5,
    '2-3 years':2.5,
    '3-4 years':3.5,
    '4-5 years':4.5,
    '5-10 years':7.5,
    '10-20 years':15,
    '20 or more years':20,
}

def get_ml_experience_int(string):
    if string is np.nan:
        return np.nan
    else:
        return ml_experience_dict[string]
```

```python
# Q21
company_size_dict = {
    '0-49 employees':24.5,
    '50-249 employees':149.5,
    '250-999 employees':624.5,
    '1000-9,999 employees':5499.5,
    '10,000 or more employees':10000,
}

def get_company_size_int(string):
    if string is np.nan:
        return np.nan
    else:
        return company_size_dict[string]
```

```python
# Q22
datascientist_dict = {
    '0':0,
    '1-2':1.5,
    '3-4':3.5,
    '5-9':7,
    '10-14':12,
    '15-19':17,
    '20+':20,
}

def get_mean_datascientist(string):
    if string is np.nan:
        return np.nan
    else:
        return datascientist_dict[string]
```

```python
# Q26
def get_mean_expenditure(string):
    if string is np.nan:
        return np.nan
    else:
        stripped = str(string).replace('$', '').replace('(USD)', '').replace(',', '').replace(' or more', '')
        splitted = stripped.split('-')
        splitted_int = [int(s) for s in splitted]
        mean_expenditure = sum(splitted_int) / len(splitted_int)
        return mean_expenditure
```

```python
def get_dummy(string):
    if string is np.nan:
        return 0
    else:
        return 1
```

```python
# Q1
df_new.loc[:, 'Q1_mean'] = df['Q1'].map(get_mean_age)
# Q2
df_new.loc[:, 'Q2_bool'] = df['Q2'].map(get_sex_bool)
# Q4
df_new.loc[:, 'Q4_int'] = df['Q4'].map(get_education_level_int)
# Q6
df_new.loc[:, 'Q6_mean'] = df['Q6'].map(get_mean_experience)
# Q13
df_new.loc[:, 'Q13_int'] = df['Q13'].map(get_tpu_experience_int)
# Q15
df_new.loc[:, 'Q15_int'] = df['Q15'].map(get_ml_experience_int)
# Q21
df_new.loc[:, 'Q21_int'] = df['Q21'].map(get_company_size_int)
# Q22
df_new.loc[:, 'Q22_mean'] = df['Q22'].map(get_mean_datascientist)
# Q26
df_new.loc[:, 'Q26_mean'] = df['Q26'].map(get_mean_expenditure)
# dummy
for dummy in dummy_list:
    columns = df.filter(like=dummy, axis=1).keys()
    for column in columns:
        df_new.loc[:, column] = df[column].map(get_dummy)
```

```python
def get_correlation_ratio(df, key):
#     x：量的変数（Q25_mean）
#     y：質的変数
    x = df['Q25_mean']
    y = df[key]
    all_mean = x.dropna().mean()
    all_var = ((x.dropna() - x.dropna().mean())**2).sum()
#     y_unique_list = get_unique_list(y)
    y_unique_list = y.unique()
    inter_var = [len(x[y==i].dropna()) * (x[y==i].dropna().mean() - all_mean)**2 for i in y_unique_list]
    r = sum(inter_var) / all_var
    return r
```

```python
def get_pearson_correlation(df, key):
    x = df['Q25_mean']
    y = df[key]
    r, _ = pearsonr(x, y)
    return r
```

```python
def get_corr(df, categorical_keys):
#     df：pandas.DataFrame
#     categorical_keys：list
    
    keys = df.keys()
    corr_ratio = pd.DataFrame({})
    corr = pd.DataFrame({})
    for key1 in keys:
        if (key1 in categorical_keys):
            r = get_correlation_ratio(df.dropna(subset=[key1,'Q25_mean']), key1)
            corr_ratio.loc[key1, 'Q25_mean'] = r
        
        elif (key1 not in categorical_keys):
            r = get_pearson_correlation(df.dropna(subset=[key1,'Q25_mean']), key1)
            corr.loc[key1, 'Q25_mean'] = r
    return corr, corr_ratio
```

```python
categorical_col = df_new.select_dtypes(include='object').columns
corr, corr_ratio = get_corr(df_new, categorical_col)
```

```python
corr_sorted = corr.sort_values(['Q25_mean'], ascending=False)
corr_sorted.style.set_caption('Correlation coefficient with compensation')
corr_sorted.head(10)
```
