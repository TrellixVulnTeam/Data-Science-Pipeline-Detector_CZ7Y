```python
data = pd.read_csv('../input/kaggle-survey-2021/kaggle_survey_2021_responses.csv')
data.head()
```

## Since I'm kaggle beginner, so I want to get info(ex.recommended language,environment, etc...) about experienced peopleðŸ˜„
### So I'll use data from who have been writing code for more than 3years

```python
data.Q6.unique()
```

```python
analy_data = data[(data.Q6=='3-5 years')|(data['Q6']=='5-10 years')|
                  (data['Q6']=='10-20 years')|(data['Q6']=='20+ years')]
analy_data.head()
```

```python
analy_data.shape
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
from plotly.offline import init_notebook_mode,iplot
import plotly.figure_factory as ff
```

```python
total = analy_data['Q6'].value_counts()
pdata= go.Pie(values=total,
              labels=total.index,
             textinfo='label+value+percent')
fig = go.Figure(data=pdata)
fig.show()
```

## First, I want to know about which tool or environment they are using.

```python
import plotly.express as px
```

## Q7(language that regular basis)

```python
#https://www.kaggle.com/sakuraandblackcat/kaggle-survey-2021-english-and?scriptVersionId=79850817&cellId=32
Q7_total=[]
for i in analy_data.columns:
    if 'Q7' in i:
        Q7_total.append(i)

Q7=analy_data[Q7_total]
Q7_sort = Q7.describe(include='all').T.sort_values('freq',ascending=False)
Q7_sort['proportion']=Q7_sort['freq']/Q7.shape[0]
```

```python
Q7_sort[['top','freq','proportion']]
```

```python
plt.figure(figsize=(11,7))
sns.barplot(x="top",y="proportion",data=Q7_sort)
plt.title('regular language')
plt.show()
```

## Q8(Recommended language)

```python
#orders=[ '3-5 years', '5-10 years','10-20 years', '20+ years' ]
sns.countplot(y='Q8',data=analy_data)
plt.title('Recommended language')
plt.show()
```

### Machine learning methods was used less than expected

```python
from plotly.subplots import make_subplots
fig = make_subplots(rows=1,cols=2,specs=[[{"type": "pie"}, {"type": "pie"}]])

total1=analy_data[analy_data['Q6']=='3-5 years']['Q15'].value_counts()
fig.add_trace(
    go.Pie(values=total1,labels=total1.index,
              textinfo='label+value+percent',
          title='3-5 years'),
    row=1,col=1)
total2=analy_data[analy_data['Q6']=='5-10 years']['Q15'].value_counts()
fig.add_trace(
    go.Pie(values=total2,labels=total2.index,
              textinfo='label+value+percent',
          title='5-10 years'),
    row=1,col=2)
fig.update_layout(title_text='years they used machine learning methods')
fig.show()
```

```python
total3=analy_data[analy_data['Q6']=='10-20 years']['Q15'].value_counts()
fig.add_trace(
    go.Pie(values=total3,labels=total3.index,
              textinfo='label+value+percent',
          title='10-20 years'),
    row=1,col=1)
total4=analy_data[analy_data['Q6']=='20+ years']['Q15'].value_counts()
fig.add_trace(
    go.Pie(values=total4,labels=total4.index,
              textinfo='label+value+percent',
          title='20+ years'),
    row=1,col=2)
fig.update_layout(title_text='years they used machine learning methods(10-20 years & 20+ years)')
fig.show()
```
