#### 1.1 Functions for Calculating Percentages <a name="func"></a>

```python
#  Defining a function that will give percentages for countplots without the hue parameter
def without_hue(plot, feature):
    total = len(feature)
    for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_height()/total)
        x = p.get_x() + p.get_width() / 2 - 0.05
        y = p.get_y() + p.get_height()
        ax.annotate(percentage, (x, y), size = 12)
    plt.show()
def without_hueh(plot, feature):
    total = len(feature)
    for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_width()/total)
        y = p.get_y() + p.get_height() / 2 - 0.05
        x = p.get_x() + p.get_width()
        ax.annotate(percentage, (x, y), size = 12)
    plt.show()
```

```python
#  Defining a function that will give percentages for countplots with the hue parameter
def with_hue(plot, feature, Number_of_categories, hue_categories):
    a = [p.get_height() for p in plot.patches]
    patch = [p for p in plot.patches]
    for i in range(Number_of_categories):
        total = feature.value_counts().values[i]
        for j in range(hue_categories):
            percentage = '{:.1f}%'.format(100 * a[(j*Number_of_categories + i)]/total)
            x = patch[(j*Number_of_categories + i)].get_x() + patch[(j*Number_of_categories + i)].get_width() / 2 - 0.15
            y = patch[(j*Number_of_categories + i)].get_y() + patch[(j*Number_of_categories + i)].get_height() 
            ax.annotate(percentage, (x, y), size = 12)
    plt.show()
```

### 2.1 Duration  <a name="dura"></a> 

```python
#  Defining a function to change duration into mintues and forming groups
kaggle_tidy=kaggle_tidy.rename(columns = {'Time from Start to Finish (seconds)':'duration'})
kaggle_tidy['duration']= kaggle_tidy['duration'].astype('int')
def duration_class(duration_grup):
    if duration_grup.duration < 120:
        return '< 2 min'
    elif duration_grup.duration < 300:
         return '2-5 min'
    elif duration_grup.duration < 600:
         return '5-10 min'
    elif duration_grup.duration < 1200:
         return '10-20 min'
    elif duration_grup.duration < 3600:
        return '20-60 min'
    else:
         return '60 min +'
        
duration_grouped= kaggle_tidy.apply(duration_class ,axis='columns') 
kaggle_tidy['duration_group'] =pd.Series(duration_grouped)
```

```python
# Question 0: Duration
plt.figure(figsize=(12,7))
# plt.xticks(rotation ='vertical')
plt.title('Time taken to complete survey', fontsize =20, bbox={'facecolor':'0.8', 'pad':5})
ax=sns.countplot(x =kaggle_tidy['duration_group'],order =['2-5 min','5-10 min', '10-20 min', '20-60 min', '60 min +']
                ,palette ='Blues', hatch='/',linewidth=1.5, edgecolor='0.2')
plt.xlabel("Time (mins)", fontsize=14,fontweight ='bold')
plt.ylabel("Number of Participants", fontsize=14,fontweight ='bold')
plt.xticks(fontsize=12)
without_hue(ax,kaggle_tidy['duration_group'])
plt.show()
```

##### Approximately 80% of the participants took the survey in the estimated time of 20 minutes. Most people (35.6%) spent between 10-20 mins to complete the survey. Below we will look at those who took over 20 mins using their age.

### 2.6 Current Title  <a name="title"></a> 

```python
# Question 5

labels =['Student','Data Scientist','Software Engineer','Other','Data Analyst','Not employed',
         'Research Scientist','ML Engineer', 'BA','PPM','DE','PM','S','DBA','DR'] 
sizes = kaggle_tidy.Q5.value_counts()
perc = [f'{i/sizes.sum()*100:5.2f}%' for i in sizes]
lbl = [f'{el[0]} = {el[1]}' for el in zip(labels, perc)]

cmap = matplotlib.cm.Blues
mini=min(sizes)
maxi=max(sizes)
norm = matplotlib.colors.Normalize(vmin=mini, vmax=maxi)
colors = [cmap(norm(value)) for value in sizes]

plt.figure(figsize=(14,8), dpi= 80)
squarify.plot(sizes=sizes, label=lbl, color=colors,alpha=1,pad = True,
              text_kwargs={'fontsize':8, 'fontname':"Trebuchet MS",'weight':'bold'})

plt.title('Treemap of Current Role', fontsize =20, bbox={'facecolor':'0.7', 'pad':5})
plt.axis('off')
plt.show()
```

```python
#  Defining a function to group salaries into smaller categories
def salary_class(salary_grup):
    if salary_grup['Q25']=='$0-999':
        return '< $1K'
    elif (salary_grup['Q25']=='1,000-1,999'or
         salary_grup['Q25']=='2,000-2,999'or
         salary_grup['Q25']=='3,000-3,999'or
         salary_grup['Q25']=='4,000-4,999'or
         salary_grup['Q25']=='5,000-7,499'or
         salary_grup['Q25']=='7,500-9,999'):
        return '1-10K'
    elif (salary_grup['Q25']=='10,000-14,999'or
          salary_grup['Q25']=='15,000-19,999'or
          salary_grup['Q25']=='20,000-24,999'):
         return '10-25K'
    elif (salary_grup['Q25']=='25,000-29,999'or
          salary_grup['Q25']=='30,000-39,999'or
          salary_grup['Q25']=='40,000-49,999'):
         return '25-50K'
    elif (salary_grup['Q25']=='50,000-59,999'or
          salary_grup['Q25']=='60,000-69,999'or
          salary_grup['Q25']=='70,000-79,999'or
          salary_grup['Q25']=='80,000-89,999'or
          salary_grup['Q25']=='90,000-99,999'):
        return '50-100K'
    elif (salary_grup['Q25']=='100,000-124,999'or
          salary_grup['Q25']=='125,000-149,999'or
          salary_grup['Q25']=='150,000-199,999'or
          salary_grup['Q25']=='200,000-249,999'):
          
        return '100-250K'
    elif (salary_grup['Q25']=='250,000-299,999'or
          salary_grup['Q25']=='300,000-499,999'):
                  
        return '250-500K'
    
    elif (salary_grup['Q25']=='$500,000-999,999'or
          salary_grup['Q25']=='>$1,000,000'):
                  
        return '>500K'
                
salary_grouped= kaggle_tidy.apply(salary_class ,axis='columns') 
kaggle_tidy['salary_group'] =pd.Series(salary_grouped)
```

  ####  1.5 Current Role <a name="titlea"></a>
     
   

```python
# Question 5

labels =['Student','Data Scientist','Data Analyst','Not employed','Other','Software Engineer',
         'ML Engineer','Research Scientist','BA', 'PPM','S','DE', 'PM','DBA','DR'] 
sizes = Afri.Q5.value_counts()
perc = [f'{i/sizes.sum()*100:5.2f}%' for i in sizes]
lbl = [f'{el[0]} = {el[1]}' for el in zip(labels, perc)]

cmap = matplotlib.cm.Blues
mini=min(sizes)
maxi=max(sizes)
norm = matplotlib.colors.Normalize(vmin=mini, vmax=maxi)
colors = [cmap(norm(value)) for value in sizes]

plt.figure(figsize=(13,7), dpi= 100)
squarify.plot(sizes=sizes, label=lbl, color=colors,alpha=0.8,pad = True,
              text_kwargs={'fontsize':9, 'fontname':"Trebuchet MS",'weight':'bold'})

plt.title('Treemap of Current Role', fontsize =20, bbox={'facecolor':'0.7', 'pad':5})
plt.axis('off')
plt.show()
```

##### Students have the highest percentage of participants (23.11%). In second place and third place are Data Scientist (13.93%) Data Analyst (12.77%) respectively. What is worrying is that unemployed participants are fourth with 10.87%. This might raise concern of high unemployement.
