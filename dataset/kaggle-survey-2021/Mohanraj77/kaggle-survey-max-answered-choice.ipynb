```python
import numpy as np 
import pandas as pd
pd.set_option('display.max_rows', None, 'display.max_columns', None,'display.max_colwidth', None)
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

## Importing data and converting columns to actual questions

```python
df1 = pd.read_csv("/kaggle/input/kaggle-survey-2021/kaggle_survey_2021_responses.csv",skiprows=[0],header=0)
p = []
for i,col in enumerate(df1.columns):
    try:
        if ((col.split('- Selected')[0]== df1.columns[i-1].split('- Selected')[0])|(col.split('- Selected')[0]== df1.columns[i+1].split('- Selected')[0])): p.append(col.split('- Selected')[0])
        else: p.append(col.split('- Selected')[0])
    except IndexError:
        if (col.split('- Selected')[0]== df1.columns[i-1].split('- Selected')[0]) : p.append(col.split('- Selected')[0])
        else: p.append(col.split('- Selected')[0])
df1.columns = p
df1.reset_index(inplace=True)
df1.head(1)
```

## Melting it and creating a column for count of each answer

```python
df2 = df1.melt(id_vars = ['index'],var_name='Ques',value_name='ans')
df3 = df2.groupby(['Ques','ans']).agg(['count']).reset_index()
df3.columns = list(map('_'.join,df3.columns))
df3.head(10)
```

## Finding the maximum answered choice for each question

```python
df4 = df3.copy()
df4['max'] = df4.groupby(['Ques_'])['index_count'].transform(lambda x : x.idxmax())
df4 = df4.reset_index()
df5 = df4.copy()
df6 = df5[df5['max']==df5['index']]
df7 = df6.drop(['index','max'],axis=1)
df7.head()
df8=df7.reset_index(drop=True)
df8
```
