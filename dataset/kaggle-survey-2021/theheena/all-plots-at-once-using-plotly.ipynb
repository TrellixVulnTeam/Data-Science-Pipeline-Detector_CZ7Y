```python
import numpy as np
import pandas as pd 
import copy
import datetime as dt
import os
import matplotlib.pyplot as plt
from plotly import tools
import plotly.offline as py
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
import plotly.offline as offline
import plotly.express as px
offline.init_notebook_mode()

pd.set_option('display.max_columns', None)
```

```python
data = pd.read_csv('../input/kaggle-survey-2021/kaggle_survey_2021_responses.csv')
questions = data.iloc[0]
data = data.drop([0])
data.reset_index(drop=True, inplace=True)
```

```python
def plot_categorical_data(cat_data, title, size=12):
    '''This funtion plots the categorical data'''
    
    values = cat_data.value_counts().sort_values(ascending=False)
    fig = px.bar(values, title=title)
    fig.update_layout(xaxis_title=None,
                      yaxis_title=None, 
                      title_x=0.5,
                      font=dict(size=size),
                      showlegend=False)
    fig.show()
    return None



def plot_multi_col_data(startwith_text, size=12):
    '''This function plots the multiple choice questions asked. It plots multi-cols data.'''
    
    multi_col_matrix = data.loc[:, data.columns.str.startswith(startwith_text)].apply(pd.Series.value_counts)
    if not multi_col_matrix.empty:
        temp_df = pd.DataFrame()
        for i, (name, values) in enumerate(multi_col_matrix.iteritems()):
            values = values.unique()
            temp_df[multi_col_matrix.index[i]] = [values for values in values if not np.isnan(values)]


        title = questions[questions.index.str.startswith(startwith_text)][0].split('?')[0]
        fig = px.bar(temp_df.T[0].sort_values(ascending=False), title=title)
        fig.update_layout(xaxis_title=None, 
                          yaxis_title=None, 
                          title_x=0.5,
                          font=dict(size=size),
                          showlegend=False)
        fig.show()
    return None
    
    
```

```python
cat_cols = ['Q1', 'Q2', 'Q4', 'Q5', 'Q6', 'Q8', 'Q11',
            'Q13', 'Q15', 'Q20', 'Q21', 'Q22', 'Q23',
            'Q25', 'Q26', 'Q28', 'Q33', 'Q35', 'Q41']

multi_cols = ['Q7', 'Q9', 'Q10', 'Q12', 'Q14', 'Q16',
              'Q17', 'Q18', 'Q19', 'Q24', 'Q27_A', 
              'Q29_A', 'Q30_A', 'Q31_A', 'Q32_A',
              'Q34_A', 'Q36_A', 'Q37_A', 'Q38_A',
              'Q39', 'Q40', 'Q42', 'Q27_B', 'Q29_B',
              'Q30_B', 'Q31_B', 'Q32_B', 'Q34_B',
              'Q36_B', 'Q37_B', 'Q38_B']
```

```python
# Plot all Categorical columns
for col in cat_cols:
    plot_categorical_data(data[col], questions[col])
```

```python
# Plot all Multi columns data
for col in multi_cols:
    plot_multi_col_data(col)
```
