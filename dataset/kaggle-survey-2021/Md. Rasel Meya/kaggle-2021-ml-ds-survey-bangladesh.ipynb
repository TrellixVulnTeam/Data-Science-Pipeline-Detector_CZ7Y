# Import Libraries

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

# Load Dataset 



```python
overall_data=pd.read_csv('../input/kaggle-survey-2021/kaggle_survey_2021_responses.csv')
```

```python
overall_data.head()
```

# Split all data for Bangladesh



```python
country='Bangladesh'
BD_data=overall_data[overall_data.Q3==country]
```

<div style="

```python
BD_data.head()
```

```python
BD_data.shape
```

# shown below pie chart for overall countries and percentage of responses

```python
fig = px.pie(overall_data, overall_data.columns[3], 
       title=f"{len(overall_data[overall_data[overall_data.columns[3]]==country])*100/len(overall_data):.2f}% of all survey respondents are from {country}", 
       hole=0.5)
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(uniformtext_minsize=10, uniformtext_mode='hide')
fig.show()
```

# Age-Gender Relationship

```python
#aggregate basis cat data for age and gender
agg_data = BD_data.groupby(['Q1','Q2']).size().reset_index(name='respondent_count')

agg_data = agg_data.rename(
    columns={
        'Q1': 'age', 
        'Q2': 'gender', 
    })
```

```python
fig = px.bar(
    agg_data, 
    x='age', 
    y='respondent_count', 
    color='gender', 
    title="Kaggle 2021 Survey Respondens by Age and Gender(Bangladesh)", 
    height=400)
fig.show()
```
