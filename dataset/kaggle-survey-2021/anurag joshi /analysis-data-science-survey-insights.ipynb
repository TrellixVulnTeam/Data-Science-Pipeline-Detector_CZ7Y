# Top 10 Kaggler woman's  Age-Group Country wise - 

### India is top most and the age-group is (18 - 21)

```python
women_kaggle = data_[data_["Q2"] == "Woman"]
```

```python
cn = women_kaggle.pivot_table(index = ["Q3", "Q1"], 
                                  values = ["Q2"], 
  aggfunc="count").reset_index().sort_values(["Q2"], ascending=False)

age_grp = cn.groupby('Q3').head(1)
age_grp.rename(columns = {"Q3" : "Country", "Q1" : "Age_Group", "Q2" : "Count_of_woman"}, inplace = True)

top_10_ = age_grp.head(10)
```

```python
sns.set(rc={'figure.figsize':(10,7.27)})
sns.scatterplot(data=top_10_, x="Age_Group", y="Count_of_woman", size="Count_of_woman", hue="Country", legend="brief", sizes=(20, 2000))
plt.legend(bbox_to_anchor=(1.51, 1),borderaxespad=0)
```
