```python
def Rusers(YearDataFrame, question):
    x = YearDataFrame.groupby(question).size()
    return x.reset_index()[:1].values[0][1]
```

# Slow Acceptance of Javascript, according to Tensorflow community

Tensorflow has a youtube channel. Now and then, I watch the videos. I haven't started with Tensorflow yet. Recently, I saw a video where they say that the Javascript community of Tensorflow has shown increase in 2020. The increase has not been much compared to Python but in a period where all the languages are hardly meeting the popularity of Python, a slight increase means a great deal for that particular language users. Let's verify the claim made by Tensorflow people.

We will only include professions that mainly deal with data. I have considered these six profiles: 'Data Scientist', 'Data Analyst', 'Research Assistant', 'Research Scientist', 'Statistician' and 'Machine Learning Engineer'.

I have created a function that calculates the number of JavaScript users and the number of respondents in these professions.
We will start calculating the number of JavaScript users from 2018 till 2021.

```python
def Jusers(year, question1, question2):
    j = year[(year[question1].isin(Datapeople))].groupby([question2]).size()[0]
    return j

def totaldatapeople(year, question):
    k = year[(year[question].isin(Datapeople))].groupby([question]).size().sum()
    return k
```

```python
Datapeople = ['Data Scientist', 'Data Analyst', 'Research Assistant', 'Research Scientist', 
              'Statistician', 'Machine Learning Engineer']
JSusers = [Jusers(df2018, 'Q6', 'Q16_Part_6'), Jusers(df2019, 'Q5', 'Q18_Part_7'), 
           Jusers(df2020, 'Q5', 'Q7_Part_7'), Jusers(df2021, 'Q5', 'Q7_Part_7')]

TotalDataPeople = [totaldatapeople(df2018, 'Q6'), totaldatapeople(df2019, 'Q5'),
                   totaldatapeople(df2020, 'Q5'), totaldatapeople(df2021, 'Q5')]

Per_JS_user = [(i / j) * 100 for i, j in zip(JSusers, TotalDataPeople)]
```

```python
TFimage = img.imread("../input/tf-imafes/1aee0ede85885520.png")

fgr = plt.figure(figsize=(8, 4))
plotting_bar = fgr.add_axes([0, 0, 1, 1])
TF = fgr.add_axes([1.1, 0.15, 0.55, 0.75])
plotting_bar.plot(year, Per_JS_user,
            color= '#2a9d8f')
plotting_bar.set_ylabel("Percentage of JavaScript users")
plotting_bar.set_xlabel("Year")
plotting_bar.set_title("Percentage of JavaScript users in  \n Machine Learning and Data Science community")
TF.imshow(TFimage)
TF.axis("off")
```

The percentage of Machine Learning JavaScript community has definitely increased in 2020. In 2021, it has not risen much but as the TF community said "it is slowly growing". It's all good as long as it does not fall. So we have confirmed that Javascript/Tensorflow JS are rising in popularity.
