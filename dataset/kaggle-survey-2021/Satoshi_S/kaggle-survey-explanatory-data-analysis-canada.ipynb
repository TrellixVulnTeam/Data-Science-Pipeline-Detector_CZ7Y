```python
def count_people(country,gender):
    country_df=df[(df.Q3==country) & (df.Q2==gender)]
    counts_country = []
    for i in age_groups: 
                try:
                    counts_country.append(country_df.Q1.value_counts()[i])
                except:
                    counts_country.append(0)
    return counts_country

def make_bargraph(country):
    # width of the bars
    barWidth = 0.3
    bars1=count_people(country,'Man')
    bars2=count_people(country,'Woman')
    # The x position of bars
    r1 = np.arange(len(bars1))
    r2 = [x + barWidth for x in r1]
 
    # Create blue bars
    plt.bar(r1, bars1, width = barWidth, color = 'blue', edgecolor = 'black', capsize=7, label='Man')
 
    # Create cyan bars
    plt.bar(r2, bars2, width = barWidth, color = 'cyan', edgecolor = 'black',capsize=7, label='Woman')
 
    # general layout
    plt.xticks([r + barWidth for r in range(len(age_groups))], age_groups)
    plt.ylabel('counts')
    plt.legend()
 
    plt.title("Kagglers Demographic in " + str(country))
    # Show graphic
    plt.show()
```

```python
countries=df.Q3.unique()[1:].tolist()
for i in countries:
    make_bargraph(i)
```

```python
def to_list(column):
    ls = df[str(column)].unique()[1:].tolist()
    counts = []
    for i in ls:
        counts.append(df[str(column)][1:].value_counts()[i])
    return counts,ls
def make_piechart(column):
    fig=plt.figure(figsize=(5,5))
    my_circle =plt.Circle((0,0),0.7,color='white')

    plt.pie(to_list(column)[0],labels = to_list(column)[1],labeldistance=2.2,autopct='%1.0f%%',pctdistance=1.9)
    p=plt.gcf()
    p.gca().add_artist(my_circle)
    plt.title(df[str(column)][0],x=0.5,y=1.5)
    plt.show()
```
