# 1.Import 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
sns.set(font_scale = 2)
KS2021 = pd.read_csv('../input/kaggle-survey-2021/kaggle_survey_2021_responses.csv')
KS2021.drop([0], inplace=True)
KS2021.replace({'United States of America' : 'USA', 
                'United Kingdom of Great Britain and Northern Ireland' : 'UK',
                'Hong Kong (S.A.R.)' : 'Hong Kong'}, inplace=True)
KS2021.rename(columns={'Q1':'Age'}, inplace=True)
KS2021.rename(columns={'Q3':'Country'}, inplace=True)
```

# 2. Age and Programming Education

```python
CountryCount = KS2021.groupby('Country').count()
CountryCount = CountryCount.sort_values('Age', ascending=False)[:10]
f, axes = plt.subplots(1,1, figsize=(20, 10), sharex=False)
f = sns.barplot(data=CountryCount, x =CountryCount.index, y='Time from Start to Finish (seconds)')
f.set_ylabel("Count")
```

The figure above shows the top 10 countries with the most Kaggler.  
Of these 10 countries, Let's take a look at the data of India, Russia, the United Kingdom, which are the representative countries that are incorporating programming into primary education, and Japan, which made programming education compulsory in 2020.

```python
Age = KS2021.groupby('Age').count()
AgeJapan = KS2021[KS2021['Country']=='Japan'].groupby('Age').count()
AgeIndia = KS2021[KS2021['Country']=='India'].groupby('Age').count()
AgeRussia = KS2021[KS2021['Country']=='Russia'].groupby('Age').count()
AgeUK = KS2021[KS2021['Country']=='UK'].groupby('Age').count()

f, axes = plt.subplots(1,4, figsize=(25, 10), sharex=True)
f.tight_layout()
f = sns.barplot(data=AgeIndia, x=AgeIndia.index,y='Time from Start to Finish (seconds)', ax=axes[0])
f.set_title('India')
f.set_ylabel("Count")
f = sns.barplot(data=AgeRussia, x=AgeRussia.index,y='Time from Start to Finish (seconds)', ax=axes[1])
f.set_title('Russia')
f.set_ylabel("Count")
f = sns.barplot(data=AgeUK, x=AgeUK.index,y='Time from Start to Finish (seconds)', ax=axes[2])
f.set_title('UK')
f.set_ylabel("Count")
f = sns.barplot(data=AgeJapan, x=AgeJapan.index,y='Time from Start to Finish (seconds)', ax=axes[3])
f.set_title('Japan')
f.set_ylabel("Count")
f.set_xticks([0, 2, 4, 6, 8, 10])
```
