# Cleaning Data / Null Values üìÑ

```python
# Check null values
null_values = survey_data.isnull().sum()
null_values
```

# Data Analysis, Visualization & Histograms

```python
plt.figure(figsize=(18,8))
ax = sns.countplot(survey_data['Q5'], order = survey_data['Q5'].value_counts().iloc[:15].index)
ax.set(title = 'Current Role', xlabel = '', ylabel = '')
for item in ax.get_xticklabels():
    item.set_rotation(90)
plt.show()
```

# **Data Scientists Age**

```python
data_scientist_age = survey_data[survey_data['Q5']=='Data Scientist']
fig = px.histogram(data_scientist_age, x="Q1", title = 'Data Scientists Age').update_xaxes(categoryorder='total descending')
fig.show()
```

# Data Scientists Gender

```python
data_scientist_gender = survey_data[survey_data['Q5']=='Data Scientist']
fig = px.histogram(data_scientist_gender, x="Q2", title = 'Data Scientists Gender').update_xaxes(categoryorder='total descending')
fig.show()
```

```python
data_scientist_cert = survey_data[survey_data['Q5']=='Data Scientist']
fig = px.histogram(data_scientist_cert, x="Q4", title = 'Data Scientists Education').update_xaxes(categoryorder='total descending')
fig.show()
```
