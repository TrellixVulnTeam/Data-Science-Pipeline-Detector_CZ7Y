# Conclusions for New Entrants

These charts have shown us that the vast majority of new entrants to the field (those with less than 4 years of experience) use python as their main language and recommend that others learn it first as well.

Within Python, the main packages used for visualisation are Matplotlib and Seaborn and the main machine learning package is SciKitLearn.

These are all standard packages and are widely recommended in articles and videos as good places to start the data science journey (along with numpy and pandas). Therefore the next question is are these packages used because they are best for beginners or are they used because they are the best for industry?

To answer this we will compare the usage rates of new entrants with those who have been in the industry for longer. If the packages are the best overall, we would expect similar rates (null hypothesis), if other packages are better for more difficult problems you would expect usage rates to increase with experience (alternative hypothesis).

```python
columns = ['Python','R', 'SQL', 'C', 'C++', 'Java', 'Javascript', 'Julia', 'Swift', 'Bash', 'MATLAB',
              'No_Languages', 'Other_Language','Vis_Matplotlib', 'Vis_Seaborn', 'Vis_Plotly', 'Vis_Altair', 
              'Vis_Bokeh', 'Vis_Geoplotlib', 'Vis_Folium','ML_SciKitLearn', 'ML_TensorFlow', 'ML_Keras', 
              'ML_Pytorch', 'ML_Fast.ai', 'ML_MXNet', 'ML_XGBoost', 'ML_LightGBM', 'ML_CatBoost', 
              'ML_Prophet', 'ML_H2O 3', 'ML_Jax', 'ML_PYLightning', 'ML_Huggingface','Salary']

rf_df = df[columns].copy(deep=True).dropna()

columns.pop(-1)

y = rf_df['Salary'].values
x = rf_df[columns].values

forest = rf(max_features = 10, n_estimators = 500, random_state = 84)
forest.fit(x,y)

forest.score(x,y)
```

Putting it through a random forest and testing it on the data it has already seen (without any test-train split) only gives a score of 0.63. This shows that there is no further information to be seen in this direction
