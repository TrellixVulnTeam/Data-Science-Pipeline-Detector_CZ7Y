# Q7 プログラミング言語

**霊夢:プログラミング言語は何を使ってるんだろう？**

**Reimu:What programming language are they using?**

```python
plt.rcParams["figure.figsize"] = (12, 12)
columns = [f"Q7_Part_{x+1}" for x in range(12)]+["Q7_OTHER"]
list_name = [df_main2[columns[x]].dropna().unique()[0] for x in range(13)]
temp1 = df_young[columns].rename(columns={columns[x]: list_name[x] for x in range(13)}).count()
temp2 = df_others[columns].rename(columns={columns[x]: list_name[x] for x in range(13)}).count()
temp1 = temp1/len(df_young)
temp2 = temp2/len(df_others)

y1 = temp1.values*100
y2 = temp2.values*100

labels = list_name
width = 0.35  # the width of the bars

x = np.arange(len(labels))  

rects1 = plt.bar(x -width/2, y1, width, label='young', color="red")
rects2 = plt.bar(x + width/2, y2, width, label='others', color="gray")

plt.ylabel("Percentage [%]")
plt.xticks(x,labels)
plt.legend()
plt.title(df_main.loc[0,columns[0]])
plt.show()
```

# Q9 統合開発環境

**霊夢:youngとothersで統合開発環境は変わるのかな？**

**Reimu:Does the integrated development environment change between young and others?**

```python
plt.rcParams["figure.figsize"] = (20, 15)
plt.rcParams['font.size']=10
columns = [f"Q9_Part_{x+1}" for x in range(12)]+["Q9_OTHER"]
list_name = [df_main2[columns[x]].dropna().unique()[0] for x in range(13)]
for x in range(len(list_name)):
    if "(" in list_name[x]:
        idx = list_name[x].find("(")
        list_name[x] = list_name[x][:idx]
temp1 = df_young[columns].rename(columns={columns[x]: list_name[x] for x in range(13)}).count()
temp2 = df_others[columns].rename(columns={columns[x]: list_name[x] for x in range(13)}).count()
temp1 = temp1/len(df_young)
temp2 = temp2/len(df_others)

y1 = temp1.values*100
y2 = temp2.values*100

labels = list_name
width = 0.35  # the width of the bars

x = np.arange(len(labels))  

rects1 = plt.barh(x -width/2, y1, width, label='young', color="red")
rects2 = plt.barh(x + width/2, y2, width, label='others', color="gray")

plt.gca().invert_yaxis()
plt.xlabel("Percentage [%]")
plt.yticks(x,labels)
plt.legend()
idx = df_main.loc[0,columns[0]].find("?")
plt.title(df_main.loc[0,columns[0]][:idx+1])
plt.show()
```

# Q42 調べる時に使うメディア

**霊夢:調べる時に何を使うんだろう？**

**Reimu:What media do you use when you look up?**

```python
plt.rcParams["figure.figsize"] = (20, 15)
plt.rcParams['font.size']=10
columns = [f"Q42_Part_{x+1}" for x in range(11)]+["Q42_OTHER"]
list_name = [df_main2[columns[x]].dropna().unique()[0] for x in range(12)]
for x in range(len(list_name)):
    if "(" in list_name[x]:
        idx = list_name[x].find("(")
        list_name[x] = list_name[x][:idx]
        
temp1 = df_young[columns].rename(columns={columns[x]: list_name[x] for x in range(12)}).count()
temp2 = df_others[columns].rename(columns={columns[x]: list_name[x] for x in range(12)}).count()
temp1 = temp1/len(df_young[columns].dropna(axis=0,how ="all"))

temp2 = temp2/len(df_others[columns].dropna(axis=0,how ="all"))

y1 = temp1.values*100
y2 = temp2.values*100

labels = list_name
width = 0.35  # the width of the bars

x = np.arange(len(labels))  

rects1 = plt.barh(x -width/2, y1, width, label='young', color="red")
rects2 = plt.barh(x + width/2, y2, width, label='others', color="gray")

plt.gca().invert_yaxis()
plt.xlabel("Percentage [%]")
plt.yticks(x,labels)
plt.legend()
idx = df_main.loc[0,columns[0]].find("?")
plt.title(df_main.loc[0,columns[0]][:idx+1])
plt.show()
```

**魔理沙:無回答者を抜いてグラフにすると、youngもothersもkaggleとyoutubeがダントツだね。  
霊夢:youtubeは約10％の差があるね。**

**Marisa:If you exclude the non-responders and make a graph, kaggle and youtube are by far the best for both young and others.
