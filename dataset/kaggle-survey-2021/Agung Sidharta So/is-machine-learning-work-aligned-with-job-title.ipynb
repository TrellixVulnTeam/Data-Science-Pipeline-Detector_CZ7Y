# Kaggle Survey Data Exploration

My objective is to try to understand if the job title aligns with the work of the survey participants in area of Machine Learning.

First, let's explore Job Title of the survey participants

```python
ds = ds[1:] # we remove the question title
```

```python
ds['Q5'][1:].value_counts().plot(kind='barh')
```

Interesting, majority of survey participants are Students. As students are not actually a job title, I'll exclude them from this exercise. "Currently not employed" will also be excluded. Also, from the chart above, we can see that there're many Data Scientists that fill the survey followed by Software Engineers. This probably is an indication that more and more Software Engineers believe Machine Learning is skill they need to have to be "relevant". 

```python
ds_without_students = ds[(ds['Q5'] != 'Student') & (ds['Q5'] != 'Currently not employed')]
```

Let's explore how many years of experience each job title has with Machine Learning

```python
ds_without_students.groupby(["Q5", "Q15"]).size().unstack().plot(kind='barh', stacked=True, figsize=(20,10))
```

Interesting, majority of the job titles have under 1 year experience in Machine Learning including people with "Data Scientist", "Machine Learning Engineer", and "Research Scientist". I'd have thought majority of those job titles will be at very least 2-3 years of experience with Machine Learning. Still, those 3 titles have more experience in ML method compared to other job titles. Unfortunately, there's no question in the survey that asks the participants whether their experience is more related to Kaggle or hobby or hackathon, etc. or it's related to their work in their respective workplace. However, we have questions "Does your employer incorporate machine learning methods into their business ?". Let's explore that to check especially if the Software Engineers really work in company that incorporate ML.

```python
ds_without_students.groupby(["Q5", "Q23"]).size().unstack().plot(kind='barh', stacked=True, figsize=(20,10))
```

Interesting, if we zoom into Software Engineer specifically in the chart above, we can probably safely assume that almost half of them get their experience through Kaggle or something outside their workplace. On the other hand, Research Scientist, Machine Learning Engineer, and Data Scientist in majority work in company that has either started to explore or established ML methods in the company. There're other titles like Data Engineer, Data Analyst, and Product or Program / Project Manager that work in the company that have established ML models. I think that makes sense as more and more ML model is treated as Product / Project and it will need product / project manager and data engineering / analyst team around it.

# Conclusion (with Assumption)
From the data exploration, it seems job title really aligns with Machine Learning work. Majority of Machine Learning Engineers, Data Scientists, and Research Scientists work in company that has established ML methods and they also have more experience in ML methods compared to their job title peers. It's also worth to note there're more and more Software Engineers learning and practicing Machine Learning and more than half work in the company that are exploring or have established ML methods.
