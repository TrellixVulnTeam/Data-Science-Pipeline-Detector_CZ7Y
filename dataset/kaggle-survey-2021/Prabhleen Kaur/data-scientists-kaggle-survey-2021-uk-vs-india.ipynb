# Responses per Gender

```python

fig, (ax1, ax2) = plt.subplots(1, 2)
color = ["darkslategray", "grey", "darkgrey", "lightslategrey", "teal",]

label_uk = resUK['Q2'].value_counts().index
ax1.pie(resUK['Q2'].value_counts(), labels = label_uk,colors=color, autopct = "%1.1f%%", shadow=True)
plt.gcf().set_size_inches(18,12)
ax1.set_title("UK")
ax1.legend()

label_in = resIN['Q2'].value_counts().index
ax2.pie(resIN['Q2'].value_counts(), labels = label_in,colors=color, autopct = "%1.1f%%", shadow=True)
plt.gcf().set_size_inches(18,12)
ax2.set_title("INDIA")
ax2.legend()

fig.tight_layout(pad=3.0)

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1)

label_uk = resUK['Q4'].value_counts().index
edu_uk = resUK['Q4'].value_counts()
ax1.barh(label_uk, edu_uk, color='dimgray')
plt.gcf().set_size_inches(18,12)
ax1.set_title("UK")

label_in = resIN['Q4'].value_counts().index
edu_in = resIN['Q4'].value_counts()
ax2.barh(label_in, edu_in, color='dimgray')
plt.gcf().set_size_inches(18,12)
ax2.set_title("INDIA")

fig.tight_layout(pad=5.0)

plt.show()
```
