```python
import matplotlib.pyplot as plt
import seaborn as sns

import warnings 
warnings.filterwarnings('ignore')
```

```python
survey = pd.read_csv("../input/kaggle-survey-2021/kaggle_survey_2021_responses.csv")
survey = survey.iloc[1:, :]
```

```python
continents = pd.read_csv("../input/2020-kaggle-survey-supporting/2020 Continents.csv")
continents_dict = {k:v for k,v in zip(continents['Country'], continents['Continents'])}
survey['Continents'] = survey['Q3'].map(continents_dict)
```

```python
country_long_name_dict = {"United States of America" : "USA",
                         "United Kingdom of Great Britain and Northern Ireland" : "UK"}
survey['Q3'] = survey["Q3"].replace(country_long_name_dict)
```

```python
continent_counts = {}
for i in range(1,13):
    continent_counts.update(dict(survey[f"Q7_Part_{i}"].value_counts()))
continent_counts.update(dict(survey["Q7_OTHER"].value_counts()))
continent_counts = pd.DataFrame(continent_counts.items(), columns = ['Language', 'Counts'])
```

# **1. World**

```python
fig = plt.figure(figsize=(10,8))

colormap = ['salmon', 'cadetblue', 'lightgreen', 'mediumturquoise', 'lightgray',
            'mediumseagreen', 'gainsboro', 'orchid', 'mediumaquamarine', 'lightslategrey', 'aliceblue', 
            'palevioletred', 'steelblue']
plt.text( 0, 1.2,'World', fontsize=18, fontweight='bold', fontfamily='serif', horizontalalignment='center')
plt.pie(x=continent_counts['Counts'], wedgeprops=dict(width=0.2), autopct = '%1.0f%%',
        colors = colormap, shadow=True)
plt.legend(continent_counts['Language'], loc='lower center', bbox_to_anchor=(-1, 1))
```

* **33% Kagglers use Python on regular basis followed by 16% uses SQL.**
