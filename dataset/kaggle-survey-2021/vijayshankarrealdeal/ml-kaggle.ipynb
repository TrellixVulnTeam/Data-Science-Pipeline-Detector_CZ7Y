```python
def get_count(question_num, parts):
    questions = []
    questions = ['Q'+ str(question_num) +'_Part_'+ str(j) for j in range(1, parts)]
    questions.append('Q'+ str(question_num) + '_OTHER')
    
    categories = []
    values = []
    for i in questions:
        category = df[i].value_counts().index[0]
        val = df[i].value_counts()[0]
        
        categories.append(category)
        values.append(val)
       
    combined_df = pd.DataFrame()
    combined_df['Category'] = categories
    combined_df['Value'] = values
    
    combined_df = combined_df.sort_values(['Value'],ascending=False)
    
    return combined_df

def plot_donut(data, title, colors):
    plt.figure(figsize=(20,15))
    
    circle = plt.Circle((0,0), 0.7, color='grey')
    plt.rcParams['text.color'] = 'black'
    
    plt.pie(data['Value'], labels=data['Category'], colors=colors)
    
    p = plt.gcf()
    p.gca().add_artist(circle)
    
    plt.title(title, size=25)
    plt.show()

programming_lang = get_count(7,12)
plot_donut(programming_lang, 'Programming Languages widely used',Moonrise7_5.hex_colors)
```

```python
ide = get_count(9,12)
plot_donut(ide, 'Development environments widely used',Zissou_5.hex_colors)
```

```python
bigdata = pd.concat([y, x], axis=1)
bigdata = bigdata.reset_index(drop=True)
```

```python
def compute_age(x):
    if x.split('+')[-1] == '':
        x = x.split('+')
        x = x[0]
    else:
        x = x.split('-')
        x = (float(x[0]) + float(x[1]))/2
    return x
    
bigdata.age = bigdata.age.apply(lambda x : compute_age(x))
```

```python
bigdata.yearsofwritingcode.unique()
```

```python
def yr_of_code(x):
    if x == 'I have never written code':
        x = 0
    else:
        x = x.split('years')
        x = x[0]
        if x.split('+')[-1] == ' ':
            x = x.split('+')
            x = x[0]
        elif x.split('<')[0] == '':
            x = x.split('<')
            x = float(x[1])
        else:
            x = x.split('-')
            x = (float(x[0]) + float(x[1]))/2
    return x
```

```python
bigdata.yearsofwritingcode=bigdata.yearsofwritingcode.apply(yr_of_code)
```

```python
bigdata
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
X = sc.fit_transform(bigdata.values)
```

```python
from sklearn.cluster import KMeans    
```

```python
wss = []
for i in range(1,12):
    cluster = KMeans(n_clusters=i)
    cluster.fit_transform(X)
    wss.append(cluster.inertia_)
plt.plot(range(1,12),wss,marker = 'o')
```

```python
cluster = KMeans(n_clusters=6)
predict = cluster.fit_predict(X)
```

```python
bigdata['predict'] = predict
```
