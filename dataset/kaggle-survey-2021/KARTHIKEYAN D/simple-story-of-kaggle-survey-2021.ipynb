# Age Group InSight

```python
age_grp = survey_data['Q1'].iloc[1:].sort_values(ascending=True)
age_grp.head()
```

```python
import plotly.express as px
```

```python
age_grp.sort_values(ascending=True)

fig = px.histogram(survey_data,x=age_grp,color=age_grp)
fig.update_layout(title='Age Group Insight',xaxis_title='Age Group',plot_bgcolor='#2d3035', paper_bgcolor='#2d3035',
                  title_font=dict(size=25, color='#a5a7ab', family="Lato, sans-serif"),
                  font=dict(color='#8a8d93'),)
fig.update_yaxes(showgrid=False)
fig.show()
```

# Experience Group InSight

```python
experience_df = survey_data['Q6'].iloc[1:].sort_values(ascending=True)

experience_fig = px.histogram(survey_data,experience_df,color=experience_df)

experience_fig.update_layout(title='Experience Insight',xaxis_title='Experience Group',plot_bgcolor='#2d3035', paper_bgcolor='#2d3035',
                  title_font=dict(size=25, color='#a5a7ab', family="Lato, sans-serif"),
                   font=dict(color='#8a8d93'),)

experience_fig.update_yaxes(showgrid=False)
experience_fig.show()
```

# IDE InSight

```python
programming_df = survey_data['Q8'].iloc[1:].dropna()

programming_fig = px.histogram(survey_data,programming_df,color=programming_df)

experience_fig.update_layout(title='Programming Insight',xaxis_title='Programming',plot_bgcolor='#2d3035', paper_bgcolor='#2d3035',
                  title_font=dict(size=25, color='#a5a7ab', family="Lato, sans-serif"),
                   font=dict(color='#8a8d93'),)
programming_fig.update_yaxes(showgrid=False)
programming_fig.show()
```

```python
currentrole_df = survey_data['Q5'].iloc[1:].dropna()
currentrole_df.sort_values(ascending=False,inplace=True)

currentrole_fig = px.histogram(survey_data,currentrole_df,color=currentrole_df)

currentrole_fig.update_layout(title='Current Role Insight',xaxis_title='Current Role',plot_bgcolor='#2d3035', paper_bgcolor='#2d3035',
                  title_font=dict(size=25, color='#a5a7ab', family="Lato, sans-serif"),
                   font=dict(color='#8a8d93'),)
currentrole_fig.update_yaxes(showgrid=False)
currentrole_fig.show()
```
