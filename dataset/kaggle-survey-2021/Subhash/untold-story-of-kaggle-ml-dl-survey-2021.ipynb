<div style="background:#fad6a5; padding:1rem">
    
1. In Kaggle 2021 survey total **_42 questions_** (with their subparts) were asked.
2. Survey got ***25,973 responses*** - after excluding:
    1. _All responses with fewer than 15 answer choices_
    2. _Respondents who spent less than 2 minutes to complete the survey_
    3. _Spam and Duplicate responses flagged by survey system_
3. All questions were either multiple-choice questions (MCQ) or multi-select questions (MSQ).
4. Not all the questions were asked to all the respondents. Some of the questions were only asked students, and who are currently unemployed or/and never spent money on any cloud computing platform - All Non-professionals.
5. If a country or territory received less than 50 respondents, the survey team grouped them into a group named "Other" for the sake of anonymity.

    
<hr>
    
- <span style="font-weight:bold; color:red">Terms used and their meaning:</span>
    
    - `MCQ`: Multiple choice question - Question with only one answer
    - `MSQ`: Multiple select question - Question with one or more answers
    - `DS`: Data Science
    - `Aggregated Dataframe`: Meaning a Dataframe is first using *describe method* for categorical data and then getting the relative frequency of each category.
    
> <span style="font-size:3rem; color:red">Note:</span> <span style="font-size:1.5rem; font-weight:bold"> In my analysis, percentage share is taken only in those respondents who attempted (at least one option chosen) the question instead of complete data. For example, in the language known question (Q7), only 24941 respondents attempted the question out of all 25973 respondents, so I have taken the percentage from 24941 instead of 25973.</span>

## <span style="color:#E57F84; padding: 0 4rem; text-transform: capitalize">Helper Functions</span>

```python
## Seaborn bar plot
def plot_barchart(df, col, order=None, 
                  x_label=None, y1_label=None,
                  y2_label=None, title=None, 
                  figsize=(15,8), 
                  plot_percentage=False,
                  display_table=True, 
                  barh=False, rotation=0):
    """
    Plot a seaborn bar-plot of `col`.
    if plot_percentage = True, then it plots dual axis plot. One axis for bar plot and 
    second axis for cumulative relative frequency plot.
    
    display_table = True, displays a pandas styled table with count and frequency
    barh: True, then it plots horizontal bar plot - Horizontal bar plot does not plot the
          cumulative relative frequency.
    
    Note: Dataframe must not be aggregated for this function to plot a seaborn bar plot
    """
    
    fig, axs = plt.subplots(figsize=figsize)
    
    if barh: # Horizontal barplot
        g = df[col].value_counts(ascending=True).plot.barh(figsize=figsize, color=fresh_air)
        for i,rect in enumerate(g.patches):
            h = rect.get_height()
            w = rect.get_width()
            y = rect.get_y()
            g.annotate(f"{w}", (w, y+h/2), color='k', size=12, ha='left', va='center')
    else: # vertical barplot
        g = sns.countplot(data=df, x=col, order=order, color=fresh_air, ax=axs)
        for rect in g.patches:
            h = rect.get_height()
            w = rect.get_width()
            x = rect.get_x()
            axs.annotate(f"{h}", (x+w/2, h), color='k', size=12, ha='center', va='bottom')


    g.set_ylabel(y1_label, fontsize=14)
    g.set_xlabel(x_label, fontsize=14)
    plt.xticks(rotation=rotation)
    plt.title(title, fontdict={'size':18, 'color':midnight_blue})
    
    # create sub dataframe of the column of 'total' & 'percentage'
    sub_df = df[col].value_counts().to_frame()
    sub_df = sub_df.rename(columns={col:'Total'})
    if order:
        sub_df = sub_df.loc[order, :]
    sub_df['Percentage'] = (sub_df['Total']/sub_df['Total'].sum())*100
    sub_df = sub_df.round(2)

    # cumulative percentage line plot for vertical barplot
    if plot_percentage and not barh:
        ax2 = g.twinx()

        p = g.add_child_axes(ax2)
        p.plot(sub_df['Percentage'].cumsum(), marker='o', color='darkorange')
        p.set_ylabel(y2_label, fontsize=14)
        for x,y in enumerate(sub_df['Percentage'].cumsum()):
            p.annotate(f"{y:.2f}%", (x,y+2), ha='center', va='bottom', color='k')

    plt.show()

    if display_table:
        display(sub_df.style.bar(color=fresh_air))
```

```python
## bar plot with pandas series
def plot_pandas_series_bar(series,
                           xlabel=None, 
                           ylabel=None, 
                           title = None,
                           rotation=0,
                           figsize=(15,8)):
    """
    This function plot a bar plot of aggregated pandas series.
    
    series: pandas series (example: df['column'] or df.column)
    barh: if True, then it plots horizontal bar plot
    plot_pct: if True, then it plots a dual axis plot with cumulative relative 
    frequency/percentage. Only possible when barh=False
    
    """
    g = series.plot.bar(figsize=figsize, color=fresh_air, rot=0)
        
    g.set_xlabel(xlabel, fontsize=14)
    g.set_ylabel(ylabel, fontsize=14)
    plt.xticks(rotation=rotation)
    g.set_title(title, fontdict={'size':18, 'color':midnight_blue})

    for rect in g.patches:
        h = rect.get_height()
        w = rect.get_width()
        x = rect.get_x()
        y = rect.get_y()

        g.annotate(f"{h}", (x+w/2, h), color='k', size=12, ha='center', va='bottom')
    
    plt.show()
```

```python
## plot lineplot for pandas aggregated series

def plot_percentage_series(series, offsets=0.5, 
                           xlabel=None, ylabel=None, 
                           title=None, rotation=0):
    """
    Plot line chart with pandas series.
    """
    p = plt.plot(series, color='darkorange', marker="o")

    for line in p:
        x = line.get_xdata()
        y = line.get_ydata()
        for x1,y1 in zip(x,y):
            plt.annotate(f"{y1:.2f}%", (x1,y1+offsets), ha='center', 
                         va='bottom',color='k', fontsize=12)

    plt.xticks(rotation=rotation)
    plt.xlabel(xlabel, fontsize=14)
    plt.ylabel(ylabel, fontsize=14)
    plt.grid(linestyle='--', color=baby_blue)
    plt.title(title, fontdict={'size':18, 'color':midnight_blue})
    
    plt.show()
```

```python
def get_multiselect_df(df, like=None, sort=False, ascending=True):
    """
    Get full data and aggregated data of a multi-select column or MSQ
    
    like: str - to filter from dataframe
    __________
    Returns:
    
    data : Full dataframe of MSQ after dropping all instances full of nan values
    agg_data: Above data's freqency and percentage share in aggregated dataframe
    """
    data = df.filter(like=like)
    # drop instances with all null values
    data = data.loc[~(data.isna().sum(axis=1)==data.shape[1])]
    # aggregate categorical data
    agg_data = data.describe(include='O').T[['freq', 'top']].set_index('top')
    # add percentage feature ---------------------------------------
    agg_data['percentage'] = agg_data.freq / data.shape[0] * 100
    
    # re-arrange indexes
    if 'None' in agg_data.index:
        indexes = agg_data.drop(['Other', 'None'], 0).index.to_list() + ['Other'] + ['None']
    else:
        indexes = agg_data.drop(['Other'], 0).index.to_list() + ['Other']
        
    agg_data = agg_data.loc[indexes]
    
    if sort:
        ## excluding "None" and "Other" indexes
        if 'None' in agg_data.index:
            indexes = (agg_data.drop(['Other', 'None'], 0)
                       .sort_values(by='freq', ascending=ascending)
                       .index.to_list())
            
            agg_data = agg_data.loc[indexes + ['Other', 'None']]
            
        else:
            indexes = (agg_data.drop(['Other'], 0)
                       .sort_values(by='freq', ascending=ascending)
                       .index.to_list())
            
            agg_data = agg_data.loc[indexes + ['Other']]
            
    
    return data, agg_data
```

```python
def get_gender_comparison_table(data, index=None, column='Q2'):
    """
    This function returns a pandas style table for comparison of male and female gender.
    
    Output table contains man and woman columns with frequency and percentage/relative frequency.
    In final pandas style table red(coral) color highlights maximum value in each column and
    baby blue color highlights minimum value in each column
    """
    data1 = pd.crosstab(index=data[index], columns=data[column])
    data1.columns = [f"{col}_count" for col in data1.columns]
    data2 = pd.crosstab(index=data[index], columns=data[column], normalize='index')
    data2.columns = [f"{col}_pct" for col in data2.columns]
    
    final_data = pd.concat([data1, data2], axis=1)
    final_data = final_data.sort_values(by=final_data.columns.to_list()[:2], ascending=False)
    ## pandas style dataframe
    table = final_data.style.highlight_max(color='orange').highlight_min(color=fresh_air)
    return table
```

```python
def get_more_than_one_df(df, sort=False, ascending=False):
    """
    Get aggregated dataframe if user selected more than one choice in MSQ's.
    """
    data = df.loc[df.notna().sum(axis=1)>1]
    # total instances
    total = data.shape[0]
    # aggregate data
    data = data.describe(include="O").T
    data = data[['top', 'freq']].dropna()
    data = data.set_index('top')
    data['percentage'] = data['freq'] / total * 100

    if sort:
        indexes = data.drop("Other",0).sort_values(by="freq", ascending=ascending).index.to_list() + ['Other']
        data = data.loc[indexes]
    
    return data
```

```python
def get_number_of_choices(df, pct=False):
    """
    This function returns a series of number of choices selected in a MSQ.
    In the final series zero means - option None was selected
    """
    # drop the column with "None" option
    # it is always the 2nd last column
    data = df.drop(df.columns[-2], 1)
    data = data.notna().sum(axis=1).value_counts()
    data = data.sort_index()
    
    ## to get percentage series
    if pct:
        data = data / data.sum() * 100
    return data
```

```python
def get_separate_male_female_df(df, like=None):
    """
    This function returns separate male and female dataframe for a MSQ
    
    df : only male and female dataframe
    like: str - msq feature to filter from dataframe
    __________
    Returns:
    
    male dataframe: dataframe with only gender='Man'
    female dataframe: dataframe with only gender='Woman'
    """
    only_male_df = df.loc[df.Q2=="Man"].filter(like=like)
    only_male_df = only_male_df.loc[~(only_male_df.isna().sum(axis=1)==only_male_df.shape[1])]
    
    only_female_df = df.loc[df.Q2=="Woman"].filter(like=like)
    only_female_df = only_female_df.loc[~(only_female_df.isna().sum(axis=1)==only_female_df.shape[1])]
    
    return only_male_df, only_female_df



def get_frequency_df(df, sort=True, ascending=False):
    """
    This function returns an aggregated dataframe from a full dataframe.
    """
    data = df.describe(include="O").T[['top', 'freq']]
    data = data.set_index('top')
    # get percenatge feature by dividing number of instances
    data['percentage'] = data['freq'] / df.shape[0] * 100
    
    # re-arrange indexes
    if 'None' in data.index:
        indexes = data.drop(['Other', 'None'], 0).index.to_list() + ['Other'] + ['None']
    else:
        indexes = data.drop(['Other'], 0).index.to_list() + ['Other']
        
    data = data.loc[indexes]
    
    if sort:
        if 'None' in data.index:
            indexes = (data.drop(['Other', 'None'], 0)
                       .sort_values(by='freq', ascending=ascending)
                       .index.to_list())
            data = data.loc[indexes + ['Other', 'None']]
        
        else:
            indexes = (data.drop(['Other'], 0)
                       .sort_values(by='freq', ascending=ascending)
                       .index.to_list())
            data = data.loc[indexes + ['Other']]
        
    return data



def plot_aggregated_df(df,
                       x=None,
                       title1=None, 
                       title2=None,
                       rotation=0):
    """
    This function plots bar chart and percentage line chart for aggregated dataframe
    """
    plot_pandas_series_bar(df.freq,
                           xlabel=x, 
                           ylabel='Count', 
                           title=title1, 
                           rotation=rotation)
        
    plot_percentage_series(df.percentage, 
                           xlabel=x, 
                           ylabel="Percentage",
                           title=title2,
                           rotation=rotation)
```

```python
def plot_conditional_probability(df, index, column='Q2',
                                 normalize='columns', 
                                 labels=None, xlabel=None, 
                                 title=None, colors=[fresh_air, teal_green]):
    """
    This function plots conditional probability bar plot.
    
    if normalize = 'index' then it plots P(column | index) and 
    if normalize = 'columns' then it plots P(index | column)
    """
    data = pd.crosstab(index=df[index], 
                       columns=df[column], 
                       normalize=normalize)
    data.plot.bar(color=colors)
    if labels:
        plt.legend(labels)
        
    plt.xlabel(xlabel, fontsize=14)
    plt.ylabel("Percentage", fontsize=14)
    plt.title(title, fontdict={'size':18, 'color':midnight_blue})
    
    plt.show()
```

```python
def compare_man_and_woman(df, like=None, xlabel=None, 
                          title1=None, title2=None, title3=None):
    """
    complete comparison of filtered feature.
    
    df: dataframe of male and female genders only
    like: str - column name (it must be a MSQ)
    title1: direct relative frequency comparision of man and woman
    title2: number of choices title
    title3: more than one choices
    
    Returns:
        Three different line plot for each title.
    """

    male_df_full, female_df_full = get_separate_male_female_df(df, like=like)

    male_freq_df = get_frequency_df(male_df_full, sort=False)
    female_ferq_df = get_frequency_df(female_df_full, sort=False)

    plot_man_vs_woman_pct(woman_series=female_ferq_df.percentage,
                         man_series=male_freq_df.percentage,
                         xlabel=xlabel,title=title1, 
                          rotation=90)


    plot_man_vs_woman_pct(woman_series=get_number_of_choices(female_df_full, pct=True),
                          man_series=get_number_of_choices(male_df_full, pct=True),
                          xlabel=f"No. of {xlabel}s using", 
                          title=title2)


    plot_man_vs_woman_pct(woman_series=get_more_than_one_df(female_df_full).percentage,
                          man_series=get_more_than_one_df(male_df_full).percentage,
                          xlabel=xlabel, 
                          title=title3, 
                          rotation=90)
```

```python

class BubbleChart:
    def __init__(self, area, bubble_spacing=0):
        """
        Setup for bubble collapse.

        Parameters
        ----------
        area : array-like
            Area of the bubbles.
        bubble_spacing : float, default: 0
            Minimal spacing between bubbles after collapsing.

        Notes
        -----
        If "area" is sorted, the results might look weird.
        """
        area = np.asarray(area)
        r = np.sqrt(area / np.pi)

        self.bubble_spacing = bubble_spacing
        self.bubbles = np.ones((len(area), 4))
        self.bubbles[:, 2] = r
        self.bubbles[:, 3] = area
        self.maxstep = 2 * self.bubbles[:, 2].max() + self.bubble_spacing
        self.step_dist = self.maxstep / 2

        # calculate initial grid layout for bubbles
        length = np.ceil(np.sqrt(len(self.bubbles)))
        grid = np.arange(length) * self.maxstep
        gx, gy = np.meshgrid(grid, grid)
        self.bubbles[:, 0] = gx.flatten()[:len(self.bubbles)]
        self.bubbles[:, 1] = gy.flatten()[:len(self.bubbles)]

        self.com = self.center_of_mass()

    def center_of_mass(self):
        return np.average(
            self.bubbles[:, :2], axis=0, weights=self.bubbles[:, 3]
        )

    def center_distance(self, bubble, bubbles):
        return np.hypot(bubble[0] - bubbles[:, 0],
                        bubble[1] - bubbles[:, 1])

    def outline_distance(self, bubble, bubbles):
        center_distance = self.center_distance(bubble, bubbles)
        return center_distance - bubble[2] - \
            bubbles[:, 2] - self.bubble_spacing

    def check_collisions(self, bubble, bubbles):
        distance = self.outline_distance(bubble, bubbles)
        return len(distance[distance < 0])

    def collides_with(self, bubble, bubbles):
        distance = self.outline_distance(bubble, bubbles)
        idx_min = np.argmin(distance)
        return idx_min if type(idx_min) == np.ndarray else [idx_min]

    def collapse(self, n_iterations=50):
        """
        Move bubbles to the center of mass.

        Parameters
        ----------
        n_iterations : int, default: 50
            Number of moves to perform.
        """
        for _i in range(n_iterations):
            moves = 0
            for i in range(len(self.bubbles)):
                rest_bub = np.delete(self.bubbles, i, 0)
                # try to move directly towards the center of mass
                # direction vector from bubble to the center of mass
                dir_vec = self.com - self.bubbles[i, :2]

                # shorten direction vector to have length of 1
                dir_vec = dir_vec / np.sqrt(dir_vec.dot(dir_vec))

                # calculate new bubble position
                new_point = self.bubbles[i, :2] + dir_vec * self.step_dist
                new_bubble = np.append(new_point, self.bubbles[i, 2:4])

                # check whether new bubble collides with other bubbles
                if not self.check_collisions(new_bubble, rest_bub):
                    self.bubbles[i, :] = new_bubble
                    self.com = self.center_of_mass()
                    moves += 1
                else:
                    # try to move around a bubble that you collide with
                    # find colliding bubble
                    for colliding in self.collides_with(new_bubble, rest_bub):
                        # calculate direction vector
                        dir_vec = rest_bub[colliding, :2] - self.bubbles[i, :2]
                        dir_vec = dir_vec / np.sqrt(dir_vec.dot(dir_vec))
                        # calculate orthogonal vector
                        orth = np.array([dir_vec[1], -dir_vec[0]])
                        # test which direction to go
                        new_point1 = (self.bubbles[i, :2] + orth *
                                      self.step_dist)
                        new_point2 = (self.bubbles[i, :2] - orth *
                                      self.step_dist)
                        dist1 = self.center_distance(
                            self.com, np.array([new_point1]))
                        dist2 = self.center_distance(
                            self.com, np.array([new_point2]))
                        new_point = new_point1 if dist1 < dist2 else new_point2
                        new_bubble = np.append(new_point, self.bubbles[i, 2:4])
                        if not self.check_collisions(new_bubble, rest_bub):
                            self.bubbles[i, :] = new_bubble
                            self.com = self.center_of_mass()

            if moves / len(self.bubbles) < 0.1:
                self.step_dist = self.step_dist / 2

    def plot(self, ax, labels, colors):
        """
        Draw the bubble plot.

        Parameters
        ----------
        ax : matplotlib.axes.Axes
        labels : list
            Labels of the bubbles.
        colors : list
            Colors of the bubbles.
        """
        for i in range(len(self.bubbles)):
            circ = plt.Circle(
                self.bubbles[i, :2], self.bubbles[i, 2], color=colors[i])
            ax.add_patch(circ)
            ax.text(*self.bubbles[i, :2], labels[i],
                    horizontalalignment='center', verticalalignment='center')



            
            
def plot_bubble_chart(data, labels, colors, legend_labels, 
                      loc='upper right', title=None, legend_title=None,
                     bbox_to_anchor=(1.5,1)):
    bubble_chart = BubbleChart(area=data, bubble_spacing=0.1)

    bubble_chart.collapse()

    fig, ax = plt.subplots(subplot_kw=dict(aspect="equal"))
    bubble_chart.plot(
        ax, labels, colors)
    ax.axis("off")
    ax.relim()
    ax.autoscale_view()
    ax.set_title(title, fontdict={'size':18, 'color':midnight_blue})
    plt.legend(labels=legend_labels, loc=loc, 
               bbox_to_anchor=bbox_to_anchor, fontsize=14,
               facecolor='w', edgecolor='k',
               borderpad=0.8,
               title=legend_title, title_fontsize=16)

    plt.show()
```

```python
def compare_msq_with_mcq(full_df, created_df, created_df_col, 
                         mcq_col, mcq_groups, join_type='inner'):
    """
    This function compares MCQ with MSQ question.
    
    full_df: Dataframe with all the instances present in both MCQ and MSQ question.
    created_df: Dataframe with all the options available in the MSQ.
    created_df_col: Name of the `created_df` column
    mcq_col: Name of the MCQ column in the Dataframe
    mcq_groups: list of all unique values in the MCQ.
    _________
    
    Returns: Dataframe for each MCQ value and MSQ options
    """
    for name in mcq_groups:
        new_df = data.loc[data[mcq_col]==name].drop(mcq_col, 1)
        new_df = new_df.describe(include="O").T[['top', 'freq']]
        new_df = new_df.rename(columns={'freq':name, 'top':created_df_col})
        
        created_df = pd.merge(left=created_df, right=new_df, 
                         how=join_type, on=created_df_col)
            
    return created_df
```

```python
def compare_msq_with_msq(full_df, full_df_columns, 
                         filter_col, created_df, created_df_col):
    
    """
    This function compares MSQ with MSQ question.
    
    full_df: Dataframe with all the instances present in both MSQs.
    full_df_columns: list of all column names of MSQ which is our current question
    filter_col: Column name of the MSQ which we want to compare with current MSQ
    created_df: Dataframe with all the options available in the MSQ of 'filter_col'.
    created_df_col: Name of the `created_df` column
    _________
    
    Returns: Dataframe for each MSQ and MSQ options
    """
    
    for col in full_df_columns:
    
        new_df = full_df.filter(regex=f"{filter_col}|{col}$")
        new_df = new_df.loc[new_df[col].notna()]

        col_name = new_df[col].unique()[0]

        new_df = new_df.drop(col, 1).describe(include="O").T[['top', 'freq']]
        new_df = new_df.rename(columns={'freq':col_name})

        created_df = pd.merge(left=created_df, right=new_df, how='inner', 
                              left_on=created_df_col, right_on="top")
        
    return created_df
```

```python
def compare_mcq_with_msq(full_data, 
                         created_df, created_df_col, 
                         mcq_options, mcq_col):
    """
    This function compares MCQ with MSQ question.
    
    full_data: Dataframe with all the instances present in both MCQ and MSQ question.
    created_df: Dataframe with all the options available in the MSQ.
    created_df_col: Name of the `created_df` column
    mcq_col: Name of the MCQ column in the Dataframe
    mcq_options: list of all unique values in the MCQ.
    _________
    
    Returns: Dataframe for each MCQ value and MSQ options
    """
    
    for value in mcq_options:
        new_df = full_data.loc[full_data[mcq_col]==value]
        new_df = new_df.drop(mcq_col, 1)
        new_df = new_df.describe(include="O").T[['top', 'freq']]
        new_df = new_df.rename(columns={'top':created_df_col, 'freq':value})

        created_df = pd.merge(created_df, new_df, how='inner', on=created_df_col)
    

    return created_df.set_index(created_df_col).T.astype('int')
```
