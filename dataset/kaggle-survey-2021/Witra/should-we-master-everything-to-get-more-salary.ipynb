## 6.1. Function

```python
def sta_sum_para_to_salary(data:pd.DataFrame, sal_cat:list, col_sum_para:list,
                          col_name_sal_cat:str)->pd.DataFrame:
    '''
    This fuction creates dataframe with list of statistical properties
    as its value.
    data = dataframe of each country which has been procesed
    salary_cat = salary category
    col_sum_para = list of columns name of the summary of the selected paramters
    output = data frame of statiscal properties of a country
    '''
    df = pd.DataFrame(columns = col_sum_para[1:])
    
    for sal_class in sal_cat:
        for sum_para in col_sum_para:
            if sum_para != "sum basic" and sal_class !=0 : # skip the basic columns and salary category of 0
                select=data.loc[data[col_name_sal_cat]==sal_class]
                if select[sum_para].tolist() != [] : 
                    mean = float('{:.{prec}f}'.format(select[sum_para].mean(), prec=1))
                    std = float('{:.{prec}f}'.format(select[sum_para].std(), prec=1))
                    med = float('{:.{prec}f}'.format(select[sum_para].median(), prec=1))  
                    mode = select[sum_para].value_counts().idxmax()
                    max_ = float('{:.{prec}f}'.format(select[sum_para].max(), prec=1))
                    min_ = float('{:.{prec}f}'.format(select[sum_para].min(), prec=1))
                    #rint(type(mode))
                    statistical_prop = [mean, std, med, mode, max_, min_]
                    df.at[sal_class, sum_para] = statistical_prop
                else:
                    df.at[sal_class, sum_para] = []
                    print(f"There is no data for {sum_para} in salary category {sal_class}")
    return df #np_country
```

```python
def background_for_parameter(fig):
    '''
    This function is to paint the background of the figure according to the parameter category
    '''
    list_para_cat =[0,1,10,16,17,19,28]
    list_color_back =['royalblue', 'skyblue','royalblue','darkgray',"#FEBFB3",'#ff7f0e']
    para_cat =["Activities","Basic", "ML", "Big Data", "Source", "Learning Path"]
    for idx in range(len(list_para_cat)-1):
        fig.add_vrect(x0=list_para_cat[idx]-0.5, x1=list_para_cat[idx+1]-0.5,
                  annotation_text=para_cat[idx], annotation_position="top",
                  fillcolor=list_color_back[idx], opacity=0.1,
                  layer="below", line_width=0)
    
```

```python
def plot_para_on_sal_cat(df_country:pd.DataFrame,country:str):
    '''
    This function plots the statistical properties on each salary category
    df_country = data frame contains the list of statiscal properties,
    country = list of considered country
    '''
    
    # Plot based on Salary category
    sal_cat = ["0 - 25K", "25K - 100K", "100K - 500K", ">500K"]
    subplot_titles = ["Just to fill 0 idx",f"Salary category 1 ({sal_cat[0]})", f"Salary category 2 ({sal_cat[1]})", 
                      f"Salary category 3 ({sal_cat[2]})", f"Salary category 4 ({sal_cat[3]})"]
    
    # check the data availablity each salary catefory
    row=[]
    for i in range(1,5): # the number of class 4 while the name is started from 1
        try:
            list_sal_group = df_country.loc[i,df_country.columns[0]:df_country.columns[-1]]
            test = [item[0] for item in list_sal_group]
            row.append(i)
        except:
            print(f"There is no salary category {i} in {country}")
    
    subplot_titles = [subplot_titles[i] for i in row]

    fig = make_subplots(len(row), 1, shared_xaxes=True, subplot_titles=subplot_titles)
    fig.update_layout(title=f'<b>Skill inspection according to Salary Category in {country}<b>',
                      xaxis_title='<b>Paramaters<b>',
                      yaxis_title='<b>Count <b>',
                      titlefont={'size': 28, 'family':'Serif'},
                      template='simple_white',
                      #showlegend=True,
                      paper_bgcolor='#edeeee',
                      plot_bgcolor='#edeeee',
                      width=1000, height=3000,
                         )
    for i in range(1,5):
        columns = df_country.columns
        try:
            list_sal_group = df_country.loc[i,df_country.columns[0]:df_country.columns[-1]]  
            mean = [item[0] for item in list_sal_group]
            std = [item[1] for item in list_sal_group]
            med = [item[2] for item in list_sal_group]
            mode = [item[3] for item in list_sal_group]
            max_ = [item[4] for item in list_sal_group]
            min_ = [item[5] for item in list_sal_group]

            stat_prop = [mean, std, med, mode, max_, min_]
            stat_name = ["mean", "std", "med", "mode", "max", "min"]
            list_color =['royalblue', 'skyblue','royalblue','darkgray','green','darkgray']
            list_dash = ['solid','dot','dot','dot','solid','dot']

            # Show plot each statistical property on four salary category
            showlegend =True
            if i!=1: showlegend=False
            for idx, stat in enumerate(stat_prop): 
                fig.add_trace(go.Scatter(y=stat, x=columns, name = stat_name[idx],
                                     line=dict(color=list_color[idx], width=4, dash=list_dash[idx]),showlegend=showlegend 
                                    ), row=i, col=1)

        except:
            pass
        #Highlight area according to the data science stage category
        background_for_parameter(fig)
    fig.show()
```
