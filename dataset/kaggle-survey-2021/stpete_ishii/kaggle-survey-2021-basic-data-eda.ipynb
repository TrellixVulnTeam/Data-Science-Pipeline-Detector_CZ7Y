```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
data0=pd.read_csv('../input/kaggle-survey-2021/kaggle_survey_2021_responses.csv')
data0[0:3].T[0:10]
```

```python
print(data0.columns.tolist()[0:10])
```

# What is your age?

```python
print(data0.iloc[0,1])
age=data0.iloc[1:,1]
age1=age.value_counts()
print(age1)
print(age1.index.tolist())
```

```python
title2 = "What is your age (# years)?"
plt.figure(figsize=(6,6), dpi=80)
plt.pie(age1, labels=age1.index.tolist(),counterclock=False,startangle=0)
plt.axis('equal')
plt.title(title2, fontsize=20)
plt.show()
```
