```python
REPLAY_PATH = 'https://www.kaggleusercontent.com/episodes/7225044.json'
```

```python
import urllib.request, json 
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter 
%matplotlib notebook
```

```python
def animate(i):
    global replay, reward, board
    plt.cla()
    #plt.style.use("ggplot")
    
    selectedAgent = replay[i][0]['action']
    atari = replay[i][0]['reward'] - reward
    reward = replay[i][0]['reward']
    
    if atari > 0:
        board[
            board.shape[0] - (board[:,selectedAgent] > 0).sum() - 1,
            selectedAgent,
            1 # green
        ] = 255
    else:
        board[
            board.shape[0] - (board[:,selectedAgent] > 0).sum() - 1,
            selectedAgent,
            0 #red
        ] = 255

    plt.imshow(board)
```

```python
with urllib.request.urlopen(REPLAY_PATH) as url:
    replay = json.loads(url.read().decode())['steps']
    
board = np.zeros((60,100,3), dtype=np.uint8)
reward = 0

fig = plt.figure(figsize=(12,6))
ani = FuncAnimation(fig, animate, interval=50, frames=2000, repeat=False)

# Feel free to comment this out to just render to browser interactively
writer = PillowWriter(fps=50)  
ani.save("7225050.gif", writer=writer)  

plt.show()
```
