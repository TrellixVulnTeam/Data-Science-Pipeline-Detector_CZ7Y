```python
subs = [19281510, 19281413]
```

```python
import requests
import time

base_url = "https://www.kaggle.com/requests/EpisodeService/"
get_url = base_url + "GetEpisodeReplay"
list_url = base_url + "ListEpisodes"

scores = {}
for s in subs:
    rj = requests.post(list_url, json = {"submissionId": int(s)}).json()
    
    scores[s] = []
    for ep in rj['result']['episodes']:
        if ep['agents'][0]['submissionId'] == ep['agents'][1]['submissionId']:
            continue
        if ep['agents'][0]['submissionId'] == s:
            scores[s].append(ep['agents'][0]['updatedScore'])
        else:
            scores[s].append(ep['agents'][1]['updatedScore'])
```

```python
import numpy as np
import plotly.io as pio

data = [dict(
	mode = 'lines',
	y = np.array(s_lb)[::-1],
    name = s_id
	) for s_id, s_lb in scores.items()]

pio.show({'data':data}, validate=False)
```
