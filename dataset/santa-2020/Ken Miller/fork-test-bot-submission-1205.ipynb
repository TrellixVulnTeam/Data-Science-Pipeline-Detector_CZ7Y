```python
%%writefile sample_agent.py

import math
from scipy.stats import beta
import numpy as np
import random

last_bandit = -1
total_reward = 0

sums_of_reward = None
numbers_of_selections = None

def ucb_agent(observation, configuration):    
    global wins, trials, last_bandit, total_reward, opp_bandit, x, actual_pulls
    
    if observation.step == 0:
        wins = [1] * configuration["banditCount"]
        
        trials = [1] * configuration["banditCount"]
        
        x = [1] * configuration["banditCount"]
        
        actual_pulls = [0] * configuration["banditCount"]
        
        opp_bandit=[]
    
    if last_bandit > -1:
        
        player = int(last_bandit == observation.lastActions[1])
        
        if player==0:
            opponent=1
        else:
            opponent=0
        
        reward = observation.reward - total_reward
        total_reward += reward
        
        if reward > 0:
            wins[last_bandit] += 1
        trials[last_bandit] += 1
        actual_pulls[last_bandit] += 1
        
        #if observation.step < 1900 and reward ==1:
            #return last_bandit
        
        opp_bandit.append(observation['lastActions'][opponent])
        
        if len(opp_bandit)>15 and observation.step < 1900:
            length=len(opp_bandit)
            if opp_bandit[length-1] == opp_bandit[length-2] and opp_bandit[length-1] != opp_bandit[length-15]:
                last_bandit=opp_bandit[length-1]
                return opp_bandit[length-1]
            
    best_proba = -1
    best_agent = None
    for k in range(configuration["banditCount"]):
        #score = win / (win + loss) * 0.97 ** (win + loss)
        proba = (wins[k]/trials[k]) * 0.97 ** (actual_pulls[k])
        x[k] = proba
        
        if proba > best_proba:
            best_proba = proba
            chosen_bandit = k

    
    if chosen_bandit is None:
        chosen_bandit = random.choice(0,99)
        last_bandit = chosen_bandit
        
    last_bandit = int(chosen_bandit)
    return int(chosen_bandit)
```
