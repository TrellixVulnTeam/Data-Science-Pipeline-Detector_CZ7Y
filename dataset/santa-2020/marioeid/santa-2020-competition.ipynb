# **applying thompson samling decay=0.97**

```python
%%writefile My_thomposon_sampling.py


import math
import random
import numpy as np
win=[]
lose=[]
bandit= None
total_reward = 0
decay=0.3
def thomposon_sampling_agent(observation, configuration):    
    global win,lose,bandit,total_reward

    if observation.step == 0:
        n_bandits=configuration.banditCount
        win=[1]*n_bandits
        lose=[1]*n_bandits
    else :
        reward = observation.reward - total_reward
        total_reward += reward
        
        
        win[bandit]+=decay*reward
        lose[bandit]+=decay*(1-reward)
        
        
    samples=np.random.beta(win,lose)
    bandit=int(np.argmax(samples))
    return bandit    
        

```

# **printing the rounds**

```python
def print_rounds(file1, file2, N=5):
    env = make("mab", debug=True)

    for i in range(N):
        env.run([file1, file2])
        p1_score = env.steps[-1][0]['reward']
        p2_score = env.steps[-1][1]['reward']
        env.reset()
        print(f"Round {i+1}: {p1_score} - {p2_score}")

print_rounds("My_thomposon_sampling.py", "My_thomposon_sampling1.py")
```
