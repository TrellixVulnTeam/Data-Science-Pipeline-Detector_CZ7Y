```python
%%writefile submission.py

import random
import numpy as np

total_reward = 0
last_action = 0

def agent_001(observation, configuration):
    
    global total_reward, last_action
    action = random.randrange(configuration.banditCount)
    
    # Attempt 1: Keep playing the same machine when winning.
    if observation.step > 0:
        this_reward = observation.reward - total_reward
        total_reward = observation.reward
        if this_reward == 1:
            action = last_action
    
    last_action = action
    return action
```
