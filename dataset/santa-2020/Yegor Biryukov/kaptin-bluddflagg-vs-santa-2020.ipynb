# Oy! [Kaptin Bluddflagg](https://warhammer40k.fandom.com/wiki/Bluddflagg) is on da bridge!

```python
%%writefile kaptin_bluddflagg.py
# comment in kaggle-environments insists dat SystemRandom is shinier than Random
from random import SystemRandom


# amount of approaches to each bandit to stomp 'im a few times
APPROACHES = 10

# use shiny random ONLY!
shiny_random = SystemRandom()

# da bandits to stomp
da_bandits = []
# da bandit to stomp now
da_bandit_to_stomp = {
    "index": None,
    "times": 0
}


def kaptin_bluddflagg(servation, figuration):
    """ WAAAGH! """
    # if it's da be'innin' of WAAAGH!
    if da_bandit_to_stomp["index"] == None:
        for i in range(APPROACHES):
            da_bandits.extend(range(figuration.banditCount))
    # if da bandit is prope'ly stomp'd
    if da_bandit_to_stomp["times"] <= 0 and len(da_bandits) > 0:
        # get da new bandit
        da_bandit_to_stomp["index"] = da_bandits.pop(shiny_random.randint(0, len(da_bandits) - 1))
        da_bandit_to_stomp["times"] = figuration["episodeSteps"] // (figuration.banditCount * APPROACHES)
    # STOMP 'IM
    da_bandit_to_stomp["times"] -= 1
    return da_bandit_to_stomp["index"]
```

# Orkz is made fer' two fings! Fightin', and winnin'!
<p style="text-align: right;">Dawn of War</p>

```python
from kaggle_environments import evaluate, make, utils

results = evaluate(
                    "mab",
#                     ["kaptin_bluddflagg.py", "random"],
#                     ["kaptin_bluddflagg.py", "kaptin_bluddflagg.py"],
                    ["kaptin_bluddflagg.py", "round_robin"],
                    
                    num_episodes=10,
                    configuration={"agentExec": "LOCAL"}
                  )

won = 0
lost = 0
tie = 0
for result in results:
    if result[0] > result[1]:
        won += 1
    elif result[0] < result[1]:
        lost += 1
    else:
        tie += 1

print(f'\nwon: {won}, lost: {lost}, tie: {tie}\n')
```
