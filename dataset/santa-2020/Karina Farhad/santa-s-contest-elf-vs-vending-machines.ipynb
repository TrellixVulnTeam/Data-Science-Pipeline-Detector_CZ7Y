```python
print("Hi, elf!")
vending_machines = int(input("How many possible vending machines in the room? ")) #By condition it's 100 vending machines
time_to_try = int(input("How many times can you try until you need to go back? ")) #By condition it's 2000 times

if vending_machines <= 2 :
    probability = 0.25 #probability of work from 0 to 1. But one of the cars will not have time to give candy cane, so the probability will be from 0 to 0.25

else:
    ap = 0.5 #probability of work from 0 to 1
    percentage_of_used_machines = int(100 * ap)
    unused_machines = int(vending_machines - 1) # The vending machine from which the candy has already been taken will not have time to cook it again.
    percentage_of_unused_machines = ("%f" % (100 - unused_machines * 100 / vending_machines))
    probability = (float(percentage_of_used_machines) - float(percentage_of_unused_machines)) / 100

candy_canes = float(time_to_try)*float(probability)
candy_canes = int(round(candy_canes, -1))
print("You can collect", candy_canes, "candy canes!")
print("Congratulations!")
```

```python
#Sorry if this is someone's code â„–2, I'm new and don't know what an agent is to publish my code for a contest

import math

last_bandit = -1
total_reward = 0

sums_of_reward = None
numbers_of_selections = None

def ucb_agent(observation, configuration):    
    global sums_of_reward, numbers_of_selections, last_bandit, total_reward

    if observation.step == 0:
        numbers_of_selections = [0] * configuration["banditCount"]
        sums_of_reward = [0] * configuration["banditCount"]

    if last_bandit > -1:
        reward = observation.reward - total_reward
        sums_of_reward[last_bandit] += reward
        total_reward += reward

    bandit = 0
    max_upper_bound = 0
    for i in range(0, configuration.banditCount):
        if (numbers_of_selections[i] > 0):
            average_reward = sums_of_reward[i] / numbers_of_selections[i]
            delta_i = math.sqrt(2 * math.log(observation.step+1) / numbers_of_selections[i])
            upper_bound = average_reward + delta_i
        else:
            upper_bound = 1e400
        if upper_bound > max_upper_bound and last_bandit != i:
            max_upper_bound = upper_bound
            bandit = i
            last_bandit = bandit

    numbers_of_selections[bandit] += 1

    if bandit is None:
        bandit = 0

    return bandit
```
