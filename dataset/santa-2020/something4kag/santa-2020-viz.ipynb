## Below are 2 example agents provided for the Candy Cane Contest.

Least pulled is an example agent from https://www.kaggle.com/c/santa-2020/discussion/202224

```python
%%writefile least_pulled.py
from kaggle_environments.envs.mab.mab import Observation, Configuration

import pandas as pd
import numpy as np



class LeastPulledAgent:
    def __init__(self, configuration: Configuration):
        self.configuration = configuration
        self.bandit_pulls = [0] * self.configuration.bandit_count

    def __call__(self, observation: Observation):
        for action in observation.last_actions:
            self.bandit_pulls[action] += 1

        pending_action = 0
        for action in range(self.configuration.bandit_count):
            if self.bandit_pulls[action] < self.bandit_pulls[pending_action]:
                pending_action = action

        return pending_action
    
    
cached_agent = None


def my_agent(raw_observation, raw_configuration):
    global cached_agent
    if cached_agent is None:
        configuration = Configuration(raw_configuration)
        cached_agent = LeastPulledAgent(configuration)

    observation = Observation(raw_observation)
    return cached_agent(observation)
```
