```python
with zipfile.ZipFile("../input/movie-review-sentiment-analysis-kernels-only/test.tsv.zip") as z:
   # open the csv file in the dataset
   with z.open("test.tsv") as f:
        
      # read the dataset
      test = pd.read_csv(f, sep='\t')
        
#       # display dataset
#       print(train.head())
```

```python
X = train.Phrase
y = train.Sentiment
X_test = test.Phrase
```

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.naive_bayes import MultinomialNB
```

```python

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

y = le.fit_transform(y)
```

```python
clf = Pipeline([('vect',CountVectorizer()),('tfidf',TfidfTransformer()),('nb',MultinomialNB())])
parameters = {'vect__ngram_range':[(1,1),(1,2)],'vect__lowercase':[True, False],
             'vect__stop_words':['english'],
              'tfidf__use_idf':[True, False],'tfidf__smooth_idf':[True,False]
             }
gs = GridSearchCV(clf, parameters
                 )
gs.fit(X, y)
```

```python
gs.best_params_
```

```python
y_pred = gs.predict(X_test)
```

```python
ids= test.PhraseId
pred = pd.DataFrame(columns=['PhraseId','Sentiment'])
lst_dict = []
for i, sentiment in zip(ids,y_pred):
    lst_dict.append((i,sentiment))
```

```python
df = pd.DataFrame(lst_dict, columns=['PhraseId', 'Sentiment'])
```

```python
pd.DataFrame(df).to_csv('submission.csv', index=False)
```
