```python
import re #regular expression operations
import pandas as pd # data processing
import numpy as np # linear algebra
from bs4 import BeautifulSoup # scrape information
import nltk #natural Language Toolkit
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer #convert a collection of text documents to a matrix of token counts
from sklearn.ensemble import RandomForestClassifier #improve the predictive accuracy and control over-fitting
from sklearn.metrics import accuracy_score #machine learning
```

```python
#load train data
train_data = pd.read_csv("../input/word2vec-nlp-tutorial/labeledTrainData.tsv.zip", header=0, delimiter="\t", quoting=3)
#get current shape of the array
print(train_data.shape) 
```

```python
#returs an array of index
print(train_data.columns.values)
```

```python
#print the first row of "review" columns 
print(train_data["review"][0])
```

```python
#delete/scrap all HTML data and stop_words and convert text to lowercase
clean_train_data = []
for review in train_data["review"]:
    review = BeautifulSoup(review).get_text()
    review = re.sub("[^a-zA-Z]", " ", review)
    review = review.lower().split()
    stop_words = set(stopwords.words('english'))
    review = [word for word in review if not word in stop_words]
    clean_train_data.append(" ".join(review))
    
clean_train_data[0]
```

```python
#transformed to a sparse matrix
vectorizer = CountVectorizer(
    analyzer="word",tokenizer=None,preprocessor=None,stop_words=None,max_features=5000) 
train_data_features = vectorizer.fit_transform(clean_train_data)
train_data_features = train_data_features.toarray()
train_data_features.shape
```

```python
#improve the predictive accuracy and control over-fitting
model = RandomForestClassifier(111)
model = model.fit(train_data_features, train_data["sentiment"])
```

```python
#load test data
test_data = pd.read_csv("../input/word2vec-nlp-tutorial/testData.tsv.zip", header=0, delimiter="\t", quoting=3)
#get current shape of the array
print(train_data.shape)
```

```python
#print the first row of "review" columns 
print(test_data["review"][0])
```

```python
#delete/scrap all HTML tags data and stop_words and convert text to lowercase
clean_test_data = []
for review in test_data["review"]:
    review = BeautifulSoup(review).get_text()
    review = re.sub("[^a-zA-Z]", " ", review)
    review = review.lower().split()
    stop_words = set(stopwords.words('english'))
    review = [word for word in review if not word in stop_words]
    clean_test_data.append(" ".join(review))
    
clean_test_data[0]
```

```python
test_data_features = vectorizer.transform(clean_test_data)
test_data_features = test_data_features.toarray()
```

```python
result = model.predict(test_data_features)
output = pd.DataFrame(data={"id": test_data["id"], "sentiment": result})
output.to_csv("submission.csv", index=False, quoting=3)
```
