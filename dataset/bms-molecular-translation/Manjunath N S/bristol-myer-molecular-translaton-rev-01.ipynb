* Reading + Visualizing data

```python
## importing all required libraries 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from os import path, listdir
from PIL import Image
```

```python
##setting ploting background
sns.set_theme("notebook")
sns.set_style( "white")
#sns.set_color_codes("")
```

```python
train_image_dir ="../input/bms-molecular-translation/train"
train_label ="../input/bms-molecular-translation/train_labels.csv"
sample_submission= "../input/bms-molecular-translation/sample_submission.csv"
```

```python
## Reading data to pandas dataframe
df = pd.read_csv( train_label,sep ="," )
df.head()
```

```python
## constructing path for all images in the training set
df["path"] = df["image_id"].apply( lambda x: train_image_dir +"/"+x[0]+"/"+x[1]+"/"+x[2]+"/"+x+".png")
```

```python
## check path constructed is correct or not for few images

for x in range( 0, 10 ):
    random_file = df["path"][np.random.randint( df.shape[0] ) ]
    if( path.exists ( random_file)):
        print ("file exist..>!")
    else:
        print ( "file doesn'texists = {}".format( random_file ))
```

```python
# Let me visualize few image + respective checmical formula
def dispaly_image( image_path, caption ):
    fig = plt.figure( figsize = ( 8,8 ), dpi = 90 )
    fig.suptitle("Chemical Formula = {}".format ( caption ) )
    plt.imshow( Image.open( image_path ) )
    plt.tight_layout()
    plt.show()
```

```python
for x in range( 0, 4) :
    image_file, caption = df[["path","InChI"]].iloc[ np.random.randint( df.shape[0] ) ]
    dispaly_image ( image_file, caption )
```
