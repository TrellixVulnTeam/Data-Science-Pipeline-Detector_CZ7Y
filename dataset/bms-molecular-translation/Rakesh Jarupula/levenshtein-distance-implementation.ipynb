# BMS molecular translation

In this notebook I will implement the [Levenshtein distance](https://en.wikipedia.org/wiki/Levenshtein_distance) metric with examples. If you find this helpful please upvote.

```python
import numpy as np
```

```python
def levenshtein_distance(a, b)-> str:
    
    d = np.zeros((len(a), len(b)))
    
    for i in range(len(a)):
        d[i, 0] = i
    
    for j in range(len(b)):
        d[0, j] = j
    
    for j in range(len(b)):
        for i in range(len(a)):
            if a[i] == b[j]:
                cost = 0
            else:
                cost = 1
            d[i, j] = min(d[i-1, j]+1, d[i, j-1] + 1,d[i-1, j-1] + cost)
    return d[len(a)-1, len(b)-1]
```

```python
levenshtein_distance('kitten', 'sitting') # k, i and g
```

```python
levenshtein_distance('InChI=1S/C13H20OS/c1-9(2)8-15-13-6-5-10(3)7-12(13)11(4)14/h5-7,9,11,14H,8H2,1-4H3',
                    'InChI=C13H20OS/c1-9(2)8-15-13-6-5-10(3)7-12(13)11(4)14/h5-7,9,11,14H,8H2,1-4H3')
# Removed 1S/ in the second sequence
```
