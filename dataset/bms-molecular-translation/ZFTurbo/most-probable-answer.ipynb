```python
# coding: utf-8
__author__ = 'ZFTurbo: https://kaggle.com/zfturbo'

import pandas as pd
import numpy as np
import operator


def sort_dict_by_values(a, reverse=True):
    sorted_x = sorted(a.items(), key=operator.itemgetter(1), reverse=reverse)
    return sorted_x


def get_avg_string():
    max_length = 403
    train = pd.read_csv('../input/bms-molecular-translation/train_labels.csv')
    print('Train length: {}'.format(len(train)))
    str1 = train['InChI'].values

    distrib = dict()
    for i in range(max_length):
        distrib[i] = dict()

    lengths = []
    for zz, s in enumerate(str1):
        lengths.append(len(s))
        l1 = len(s)
        for i in range(max_length):
            if i < l1:
                char = s[i]
            else:
                char = ' '
            if char not in distrib[i]:
                distrib[i][char] = 0
            distrib[i][char] += 1
        if zz % 100000 == 0 and zz > 0:
            print('Proc: {}'.format(zz))

    lengths = np.array(lengths)
    print('Stat:', lengths.mean(), lengths.min(), lengths.max())

    best_str = ''
    for i in range(max_length):
        d = sort_dict_by_values(distrib[i], reverse=True)
        print(d[0])
        if d[0][0] == ' ':
            break
        best_str += d[0][0]

    print('Most porbable string: {}'.format(best_str))
    return best_str


if __name__ == '__main__':
    best_str = get_avg_string()
    subm = pd.read_csv('../input/bms-molecular-translation/sample_submission.csv')
    subm['InChI'] = best_str
    subm.to_csv('submission.csv', index=False)
```
