# To rdkit molecule and back!

I have an idea, that it is possible to predict molecule graph and then convert it to rdkit molecule and then create inchi.

Let's covert inchi to rdkit molecule and turn it back into inchi.

```python
!conda install -y rdkit
```

```python
import pandas as pd
from rdkit.Chem import inchi
```

```python
df = pd.read_csv("/kaggle/input/bms-molecular-translation/train_labels.csv")
```

```python
df.head()
```

```python
df["InChI"].head()
```

```python
df["InChI"].head()[0]
```

```python
mol = inchi.MolFromInchi(df["InChI"].head()[0])
mol
```

```python
inchi.MolToInchi(mol)
```

```python
inchi.MolToInchi(mol) == df["InChI"].head()[0]
```

```python
for inch in df["InChI"]:
    rdkit_inchi = inchi.MolToInchi(inchi.MolFromInchi(inch))
    assert rdkit_inchi == inch, f"gt inchi '{inch}' != rdkit_inchi '{rdkit_inchi}'"
```

So, as it seems, this approach might lead to some errors.

Still, this approach might be useful to export molecule graph via rdkit to inchi.
