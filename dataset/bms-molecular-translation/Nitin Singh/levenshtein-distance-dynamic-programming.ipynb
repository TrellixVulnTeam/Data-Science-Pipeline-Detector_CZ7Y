```python
str1 = 'abcdef'
str2 = 'azced'
```

 Levenshtein Distance : Minimum number of edits (insertion, deletion, or replacement) to tranform str1 to str2.

Watch the following video for the explanation.

https://youtu.be/We3YDTzNXEk

```python
arr = [[None for _ in range(len(str1)+1)] for _ in range(len(str2)+1)]
```

```python
for i in range(len(str2)+1):
    arr[i][0] = i

for j in range(1, len(str1)+1):
    arr[0][j] = j
    arr[1][j] = j-1
```

```python
arr
```

```python
for i in range(2, len(str2)+1):
    for j in range(1, len(str1)+1):

        if str1[j-1] == str2[i-1]:
            arr[i][j] = arr[i-1][j-1]

        else:
            arr[i][j] = min(arr[i-1][j], arr[i-1][j-1], arr[i][j-1]) + 1
```

```python
arr
```

```python
levenshtein_dist = arr[-1][-1]
```
