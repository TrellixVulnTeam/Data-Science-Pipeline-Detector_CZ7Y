```python
import pandas as pd
import numpy as np
import glob
import re

from PIL import Image

import warnings
warnings.filterwarnings("ignore")
```

```python
INPUT_PATH = '../input/bms-molecular-translation/'
train_labels = pd.read_csv(INPUT_PATH + 'train_labels.csv')
train_labels['path'] = train_labels['image_id'].apply(lambda x: 
                                    INPUT_PATH + "/train/{}/{}/{}/{}.png".format(x[0], x[1], x[2], x))
```

```python
train_labels.head()
```

```python
train_labels = train_labels.sort_values(by='InChI')
```

```python
print(train_labels['InChI'].iloc[0])
Image.open(train_labels['path'].iloc[0])
```

```python
print(train_labels['InChI'].iloc[1])
Image.open(train_labels['path'].iloc[1])
```

```python
train_labels['InChI_len'] = train_labels['InChI'].apply(len)
train_labels['InChI_SplitCount'] = train_labels['InChI'].apply(lambda x: len(x.split('/')))
train_labels = train_labels[train_labels['InChI_SplitCount'] > 3]

train_labels['InChI_1'] = train_labels['InChI'].apply(lambda x: x.split('/')[0])
train_labels['InChI_2'] = train_labels['InChI'].apply(lambda x: x.split('/')[1])
train_labels['InChI_3'] = train_labels['InChI'].apply(lambda x: x.split('/')[2])
train_labels['InChI_4'] = train_labels['InChI'].apply(lambda x: x.split('/')[3])
```
