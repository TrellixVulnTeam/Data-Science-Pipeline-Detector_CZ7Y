```python
import pandas as pd
import numpy as np
import glob
import re

from PIL import Image

import warnings
warnings.filterwarnings("ignore")
from tqdm import tqdm_notebook
import imagehash
```

```python
INPUT_PATH = '../input/bms-molecular-translation/'
train_labels = pd.read_csv(INPUT_PATH + 'train_labels.csv')
train_labels['path'] = train_labels['image_id'].apply(lambda x: 
                                    INPUT_PATH + "/train/{}/{}/{}/{}.png".format(x[0], x[1], x[2], x))

test_labels = pd.read_csv(INPUT_PATH + 'sample_submission.csv')
test_labels['path'] = test_labels['image_id'].apply(lambda x: 
                                    INPUT_PATH + "/test/{}/{}/{}/{}.png".format(x[0], x[1], x[2], x))
```

```python
def hashimg(path):
    hash = imagehash.average_hash(Image.open(path))
    return str(hash)
```

```python
hashimg(train_labels['path'].iloc[0])
```

```python
hashimg(train_labels['path'].iloc[1])
```

```python
hashs = []
for path in tqdm_notebook(train_labels['path'].iloc[:10000]):
    hashs.append(list(hashimg(path)))
hashs = np.array(hashs)
```

```python
path = test_labels['path'].iloc[120]
ids = (hashs == np.array(list(hashimg(path)))).mean(1)
idx = np.argsort(ids)[::-1]
```
