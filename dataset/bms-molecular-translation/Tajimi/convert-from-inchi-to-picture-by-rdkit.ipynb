```python
import pandas as pd
from tqdm.auto import tqdm
tqdm.pandas()

import cv2
import matplotlib.pyplot as plt
import numpy as np


from rdkit.Chem import Draw,MolFromInchi
```

# read_csv
use @yasufuminakama 's script (https://www.kaggle.com/yasufuminakama/molecular-translation-naive-baseline). Thanks!

```python
train = pd.read_csv('../input/bms-molecular-translation/train_labels.csv')

def get_train_file_path(image_id):
    return "../input/bms-molecular-translation/train/{}/{}/{}/{}.png".format(
        image_id[0], image_id[1], image_id[2], image_id 
    )

train['file_path'] = train['image_id'].progress_apply(get_train_file_path)

print(f'train.shape: {train.shape}')
display(train.head())
```

# Convert from InChI to Image

```python
for i in range(10):
    mol_data = train.iloc[i,:]
    print("image_id: "+mol_data["image_id"])
    
    mol = MolFromInchi(mol_data["InChI"])
    display(Draw.MolToImage(mol))
    Draw.MolToFile(mol,mol_data["image_id"]+".png")
    
    image = cv2.imread(mol_data["file_path"])
    plt.imshow(image)
    plt.show()
```
