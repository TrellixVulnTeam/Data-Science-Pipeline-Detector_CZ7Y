# InjuryRecord Data

```python
InjuryRecord_Data.isnull().sum()
```

```python
InjuryRecord_Data.groupby(['BodyPart']).count()['PlayerKey'].sort_values().plot(kind="bar", figsize=(15, 5), title='Count of injuries by Body Part')
plt.show()
```

```python
InjuryRecord_Data.groupby(['Surface']).count()['PlayerKey'].sort_values().plot(kind="bar", figsize=(15, 5), title='Count of injuries by Surface')
plt.show()
```

```python
InjuryRecord_Data.groupby(['Surface','BodyPart']).count()['PlayerKey'].sort_values().plot(kind="bar", figsize=(15, 5), title='Count of injuries by Surface and BodyPart')
plt.show()
```

```python
InjuryRecord_Data.groupby(['BodyPart','Surface']).count().unstack('BodyPart')
```

```python
InjuryRecord_Data.groupby(['BodyPart','Surface']).count().unstack('BodyPart')['PlayerKey'].sort_values('Surface').plot(kind= "bar" ,figsize=(15, 5), title='Count of injuries by Surface')
plt.show()
```

On "natural" turf, max. injuries occured on knees whereas max injuries on "synthetic" surface happened on Ankle.
synthetic suface shows no injury on Player heels

```python
#lets check, is there any player who had multiple injuries:
InjuryRecord_Data[InjuryRecord_Data.duplicated(['PlayerKey'])]
```

```python
InjuryRecord_Data.groupby(['Surface','BodyPart','PlayerKey']).count().unstack('BodyPart')['GameID'].sort_values('PlayerKey').plot(kind= "bar" ,figsize=(15, 5), title='Count of player injuries ')
plt.show()
```

Player 43450 had Ankle injury twice on Synthetic Surface, whereas none other player got same injury on same surface more than once.

```python
InjuryRecord_Data.groupby(['BodyPart','PlayerKey']).count().unstack('BodyPart')['GameID'].sort_values('PlayerKey').plot(kind= "bar" ,figsize=(15, 5), title='Count of player injuries ')
plt.show()
```

Player 33337(on foot) and 43540(on Ankle) got injuries twice on same body part

```python
InjuryRecord_Data.groupby(['PlayerKey']).count().sort_values('PlayerKey').plot(kind= "bar" ,figsize=(15, 5), title='Count of player injuries ')
plt.show()
```

```python
PlayList_Data[['PlayKey','PlayType']].drop_duplicates() \
    .groupby('PlayType').count()['PlayKey'] \
    .sort_values() \
    .plot(kind='barh',
         figsize=(15, 6),
          color='Teal',
         title='Number of plays provided by type')
plt.show()
```

```python
PlayList_Data[['PlayKey','RosterPosition']].drop_duplicates() \
    .groupby('RosterPosition').count()['PlayKey'] \
    .sort_values() \
    .plot(kind='barh',
         figsize=(15, 6),
          color='olive',
         title='Number of plays provided by player position')
plt.show()
```

```python
inj_detailed = InjuryRecord_Data.merge(PlayList_Data, on = ["PlayerKey","GameID","PlayKey"] , how="left")
```

```python
inj_detailed.head(5)
```

```python
inj_detailedd = inj_detailed.drop(['FieldType'],axis=1)
```

```python
#lets plt 105 players injuries w.r.t bodyparts
inj_detailed.groupby(['PlayerKey','BodyPart']).count().unstack('BodyPart')['GameID'].sort_values('PlayerKey').sort_values('Ankle').sort_values('Knee').plot(kind= "bar" ,figsize=(15, 8), title='Count of player injuries ')
plt.show()
```

```python
inj_detailed.groupby(["Surface","RosterPosition","BodyPart"]).count()['PlayerKey'].unstack("Surface").sort_values('RosterPosition').plot(kind= "bar" ,figsize=(15, 5), title='Players Injured in their respective Play positions on different Surfaces...')
plt.show()
```

```python
inj_detailed.groupby(["RosterPosition","BodyPart"]).count()['PlayerKey'].unstack('BodyPart').sort_values('Ankle').plot(kind= "bar", figsize=(15, 5), title='Players BodyPart injuries in their respective Play positions ')
plt.show()
```

```python
inj_detailed.groupby(["PlayType","Surface"]).count()['PlayerKey'].unstack('Surface').plot(kind= "bar", figsize=(15, 5), title='Players Injured in their respective Play positions on different Surfaces...')
plt.show()
```

```python
inj_detailed.groupby(["Surface","PlayType","BodyPart"]).count()['PlayerKey'].unstack("BodyPart").sort_values('Surface', ascending=False).plot(kind= "bar" ,figsize=(15, 5), title="Players Injured BodyParts in respective plays type on different turf")
plt.show()
```
