# Describe Data
There are three files provided in the dataset, as described below:

Injury Record: The injury record file in .csv format contains information on 105 lower-limb injuries that occurred during regular season games over the two seasons. Injuries can be linked to specific records in a player history using the PlayerKey, GameID, and PlayKey fields.

Play List: – The play list file contains the details for the 267,005 player-plays that make up the dataset. Each play is indexed by PlayerKey, GameID, and PlayKey fields. Details about the game and play include the player’s assigned roster position, stadium type, field type, weather, play type, position for the play, and position group.

Player Track Data: player level data that describes the location, orientation, speed, and direction of each player during a play recorded at 10 Hz (i.e. 10 observations recorded per second).

```python
injury_data=injury_data.dropna()
natural_surface=injury_data[injury_data['Surface']=='Natural']['PlayKey']
synthetic_surface=injury_data[injury_data['Surface']=='Synthetic']['PlayKey']
onnatural_surface_speed=[]
onnatural_surface_distance=[]
onnatural_surface_orientation=[]
onnatural_surface_direction=[]
for key in natural_surface:
    playkey=playertrackdata[playertrackdata['PlayKey']==key]
    onnatural_surface_speed.extend(playkey['s'])
    onnatural_surface_distance.extend(playkey['dis'])
    onnatural_surface_orientation.extend(playkey['o'])
    onnatural_surface_direction.extend(playkey['dir'])
onsynthetic_surface_speed=[]
onsynthetic_surface_distance=[]
onsynthetic_surface_orientation=[]
onsynthetic_surface_direction=[]
for key in synthetic_surface:
    playkey=playertrackdata[playertrackdata['PlayKey']==key]
    onsynthetic_surface_speed.extend(playkey['s'])
    onsynthetic_surface_distance.extend(playkey['dis'])
    onsynthetic_surface_orientation.extend(playkey['o'])
    onsynthetic_surface_direction.extend(playkey['dir'])
```

```python
noninj_natural_surface=playlist[playlist['FieldType']=='Natural']['PlayKey']
noninj_synthetic_surface=playlist[playlist['FieldType']=='Synthetic']['PlayKey']
non_inj_natural_surface=np.random.choice(noninj_natural_surface,30,replace=False)
non_inj_synthetic_surface=np.random.choice(noninj_synthetic_surface,30,replace=False)
noninj_onnatural_surface_speed=[]
noninj_onnatural_surface_distance=[]
noninj_onnatural_surface_orientation=[]
noninj_onnatural_surface_direction=[]
for key in non_inj_natural_surface:
    playkey=playertrackdata[playertrackdata['PlayKey']==key]
    noninj_onnatural_surface_speed.extend(playkey['s'])
    noninj_onnatural_surface_distance.extend(playkey['dis'])
    noninj_onnatural_surface_orientation.extend(playkey['o'])
    noninj_onnatural_surface_direction.extend(playkey['dir'])
noninj_onsynthetic_surface_speed=[]
noninj_onsynthetic_surface_distance=[]
noninj_onsynthetic_surface_orientation=[]
noninj_onsynthetic_surface_direction=[]
for key in non_inj_synthetic_surface:
    playkey=playertrackdata[playertrackdata['PlayKey']==key]
    noninj_onsynthetic_surface_speed.extend(playkey['s'])
    noninj_onsynthetic_surface_distance.extend(playkey['dis'])
    noninj_onsynthetic_surface_orientation.extend(playkey['o'])
    noninj_onsynthetic_surface_direction.extend(playkey['dir'])
```

```python
def change(list1):
    change_list1=[]
    for index in range(1,len(list1)):
        change_list1.append(list1[index]-list1[index-1])
    return change_list1
```

```python
def count_probability(bodypart,surface):
    bodypart_surface=(injury_data['BodyPart']==bodypart) & (injury_data['Surface']==surface)
    total_particular_surface=injury_data[injury_data['Surface']==surface]['PlayerKey'].count()
    prob=bodypart_surface.sum()/total_particular_surface
    return round(prob,4)
```
