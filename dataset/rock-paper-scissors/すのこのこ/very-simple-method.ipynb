If you just play randomly according to your opponent's RPS percentage, you can win rather easily.
In other words, if your choice of moves is unbalanced, you will be taken advantage of and at a disadvantage.

```python
%%writefile submission.py

import random

recs = []

def my_agent(obs, conf):
    global recs
    
    count=len(recs)
    
    if count==0:
        hand = random.randint(0, 2)
        
    else:
        recs.append(obs["lastOpponentAction"])

        hand_count = [recs.count(0), recs.count(1), recs.count(2)]
        hand_ratio = [hand_count[0], hand_count[0]+hand_count[1], count] / count

        hand_rand = random.random()
        for i,ratio in enumerate(hand_ratio):
            if hand_rand <= ratio:
                hand = i
                break
    
    return (hand+1)%3
```
