## Epsilon Greedy on 12

```python
%%writefile submission.py
import secrets

class RandomAgent():
    
    def __init__(self):
        pass
        
    def action(self, self_moves, opp_moves):
        return secrets.choice(list(range(3)))

    
class ModalModel():
    
    def __init__(self):
        pass
        
    def action(self, self_moves, opp_moves):
        ones = 0
        twos = 0
        zeros = 0
        if len(opp_moves):
            for i in opp_moves:
                if (i % 3) == 0:
                    zeros += 1
                if (i % 3) == 1:
                    ones += 1
                if (i % 3) == 2:
                    twos += 1
            values = [zeros, ones, twos]
            return (values.index(max(values)) + 1) % 3
        else:
            action = secrets.choice(list(range(3)))
        return (action + 1) % 3


class OppRotation0():
    
    def __init__(self):
        pass
    
    def action(self, self_moves, opp_moves):
        if len(opp_moves):
            return (opp_moves[-1]) % 3
        else:
            return 0


class OppRotation1():
    
    def __init__(self):
        pass
        
    def action(self, self_moves, opp_moves):
        if len(opp_moves):
            return (opp_moves[-1] + 1) % 3
        else:
            return 0



class OppRotation2():
    
    def __init__(self):
        pass
    
    def action(self, self_moves, opp_moves):
        if len(opp_moves):
            return (opp_moves[-1] + 2) % 3
        else:
            return 0
    
    
class Rotation0():
    
    def __init__(self):
        pass
        
    def action(self, self_moves, opp_moves):
        if len(self_moves):
            return (self_moves[-1]) % 3
        else:
            return 0

        
class Rotation1():
    
    def __init__(self):
        pass
    
    def action(self, self_moves, opp_moves):
        if len(self_moves):
            return (self_moves[-1] + 1) % 3
        else:
            return 0

class Rotation2():
    
    def __init__(self):
        pass
    
    def action(self, self_moves, opp_moves):
        if len(self_moves):
            return (self_moves[-1] + 2) % 3
        else:
            return 0
    
    
class LeastCommonOpponentAction():
    
    def __init__(self):
        pass
    
    def action(self, self_moves, opp_moves):
        if len(opp_moves):
            zeros = 0
            ones = 0
            twos = 0
            for move in opp_moves:
                if (move % 3) == 0:
                    zeros += 1
                elif (move % 3) == 1:
                    ones += 1
                elif (move % 3) == 2:
                    twos += 2
            move_freq = [zeros, ones, twos]
            action = move_freq.index(min(move_freq))
        else:
            action = secrets.choice(list(range(3)))
        return action
    
    
class RockIt():
    
    def __init__(self):
        pass
    
    def action(self, self_moves, opp_moves):
        return 0


class SaltAndPaper():
    
    def __init__(self):
        pass
    
    def action(self, self_moves, opp_moves):
        return 1


class Scissors():
    
    def __init__(self):
        pass
    
    def action(self, self_moves, opp_moves):
        return 2
    
    
def winner(self_move, opp_move):
    if self_move == opp_move:
        return 0
    elif (self_move == 0 and opp_move == 2):
        return 1
    elif (self_move == 2 and opp_move == 0):
        return -1
    elif (self_move > opp_move):
        return 1
    else:
        return -1
    

def special_rounder(num):
    if num <= 0:
        return 0
    else:
        return 1

class ModelAction():
    
    def __init__(self, models):
        self.models = models
        self.num_uses = [0 for i in range(len(models))]
        self.num_wins = [0 for i in range(len(models))]
        self.win_rate = [0 for i in range(len(models))]
        self.epsilon = 0.1 # This part based on epsilon greedy algorithm
        
    def update(self, model_index, win_loss):
        # print(self.num_uses[model_index])
        # print(self.num_wins[model_index])
        # print(self.win_rate[model_index])
        # print(model_index)
        # print(win_loss)
        self.num_uses[model_index] += 1
        self.num_wins[model_index] += special_rounder(win_loss[-1])
        self.win_rate[model_index] = self.num_wins[model_index]/self.num_uses[model_index]
        
    def take_action(self):
        yes_no = secrets.randbelow(10000)
        yes_no /= 10000
        if yes_no < self.epsilon:
            return self.choose_random_model()
        else:
            return self.choose_best_model()
        
    def choose_random_model(self):
        return secrets.randbelow(len(self.models))
    
    def choose_best_model(self):
        return self.win_rate.index(max(self.win_rate))

rand_agent = RandomAgent()
mode_agent = ModalModel()
opp_rotat0 = OppRotation0()
opp_rotat1 = OppRotation1()
opp_rotat2 = OppRotation2()
self_rotat0 = Rotation0()
self_rotat1 = Rotation1()
self_rotat2 = Rotation2()
anti_modal = LeastCommonOpponentAction()
rock_agent = RockIt()
paper_agent = SaltAndPaper()
scis_agent = Scissors()
# 12 models, and counting

model_list = [rand_agent, mode_agent, opp_rotat0, opp_rotat1, opp_rotat2,
              self_rotat0, self_rotat1, self_rotat2, anti_modal, 
              rock_agent, paper_agent, scis_agent
             ]

# # Other possible models:
# return current_step % 3
# on_line_neural_network
# triangle model


my_moves = list()
opp_moves = list()
wins_losses = [list() for i in model_list]

decider = ModelAction(model_list)
possible_moves = [0 for i in model_list]
# update for all possible moves every move made

index = [0]

def make_a_move(observation, configuration):
    if observation.step > (100):
        opp_moves.append(observation.lastOpponentAction)
        for i in range(len(model_list)):
            wins_losses[i].append(winner(possible_moves[i], observation.lastOpponentAction))
#         wins_losses.append(winner(my_moves[-1], opp_moves[-1]))
        for i in range(len(model_list)):
            decider.update(i, wins_losses[i])
        # print("Updated weights...")
        # print("Index:" + str(index[0]))
        index[0] = decider.take_action()
        # print(index[0])
        # print("\n")
        for i in range(len(model_list)):
            possible_moves[i] = model_list[i].action(my_moves, opp_moves)    
        curr_action = model_list[index[0]].action(my_moves, opp_moves)
        possible_moves[index[0]] = curr_action
        my_moves.append(curr_action)
        return curr_action
    else:
        # print(index[0])
        if observation.step:
            opp_moves.append(observation.lastOpponentAction)
            for i in range(len(model_list)):
                wins_losses[i].append(winner(possible_moves[i], observation.lastOpponentAction))
            for i in range(len(model_list)):
                decider.update(i, wins_losses[i])
        index[0] = (observation.step % len(model_list))
        if observation.step:
            curr_action = model_list[observation.step % len(model_list)].action(my_moves, opp_moves)
        else:
            curr_action = model_list[observation.step % len(model_list)].action(0, 0)
            for i in range(len(model_list)):
                possible_moves[i] = model_list[i].action(my_moves, opp_moves)
            possible_moves[observation.step % len(model_list)] = curr_action
        my_moves.append(curr_action)
        return curr_action
```
