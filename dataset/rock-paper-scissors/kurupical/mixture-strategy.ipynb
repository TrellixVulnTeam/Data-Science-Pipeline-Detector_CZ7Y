```python
%%writefile submission.py
import random

i = int(random.random() * 9)
def copy_opponent_agent (observation, configuration):
    global i
    hand_idx = 0
    if random.random() < 0.05:
        i += 1
    hand_idx = i % 3
    beat_scissors_strategy = [0, 0, 0, 1, 2, 2]
    beat_paper_strategy = [0, 1, 1, 2, 2, 2]
    beat_rock_strategy = [0, 0, 1, 1, 1, 2]

    hands = [beat_scissors_strategy, beat_paper_strategy, beat_rock_strategy][hand_idx]
    return random.choice(hands)
```
