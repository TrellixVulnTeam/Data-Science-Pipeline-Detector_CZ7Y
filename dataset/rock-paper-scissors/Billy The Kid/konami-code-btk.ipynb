# The Evolutionary Process

Each move list was judged based on the amount of rounds won in a 12 round bout with a Simple Markov Chain agent. This agent was my most sucessful at the time - therefore I saw it best fit we use this as our basis of success. Once each agent in the population was judged, the bottom half of the population was replaced by children from the top half of the population. Each child is mutated slightly and the process is repeated until a move list that wins all 12 rounds is found. These mutations were more likely to occur at the end of the move list rather than the beginning. This is because the beginning half of these matches seem to be more about exploring your opponent's capabilities rather than trying to win each move.
