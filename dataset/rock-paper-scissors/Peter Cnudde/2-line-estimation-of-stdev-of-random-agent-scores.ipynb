I wanted to calculate the stdev of the score of a random RPS agent versus the number of plays. This is useful for a strategy where you want to switch to random if you are far enough ahead, or if you want to know when to declare a match a tie. The expected value of the random agent is 0, but the stdev is not.

I know normal random walks are proportional to sqrt(n), but wanted to see how it behaves with choices[-1,0,1].

Here is a very simple and fast implementation. The total distance walked can be easily calculated by multiplying with a triangular matrix of ones.

```python
import numpy as np
import matplotlib.pyplot as plt

walks_simulated = 5000
N =1000

d = np.matmul(np.random.randint(3, size=(walks_simulated,N))-1,np.triu(np.ones((N,N),dtype=np.int8)))
std = np.std(d,axis=0)

plt.plot(std)
```
