JUST A SIMPLE STARTING POINT FOR A BEGINNER.
Let try to see what openent can move in return of your movment.

for example if you play rock
oppenet can Play Rock(R) ,Paper(P) and Scissor(S).
1) If R followed by R from oppenet it can be meant he is copying your move.
2) If R followed by P it can mean oppenet is just try to beat what ever you play last.
3) If R followed by S it mean oppent just consider or hoping for you to do 2.

In this notebook i will only try to handle this three sinario and it how to play aginst it.
Things to keep in mind in this notebook we just judging by last move, adding memory can help your score to improve, and alos aginst random oppenent it can fail. 


```python
#Declaring Varable.
Rock = 0;
Paper = 1;
Sic = 2;

# 0(R) can be beaten by 1(P)
# 1(P) can be beaten by 2(S)
# 2(S) can be beaten by 0(R)

beatingDic = {0:1 , 1:2, 2:0};

myLastMovH = 0;
```

```python
# predicitng is oppenet just copy your move
def isCopyingMove(openentMove , myLastMove):
    if openentMove == myLastMove:
        return True;
    return False;

# predicitng is oppenet revecing your move if you play rock he trying beat it

def IsOpententRMove( openentMove , myLastMove):
    cond1 = ((myLastMove==Rock)and(openentMove==Paper));
    cond2 = ((myLastMove==Sic)and(openentMove==Rock));
    cond3 = ((myLastMove==Paper)and(openentMove==Sic));

    return (cond1 or cond2 or cond3);

# prediciting is oppenent just hoping you do 2
def IsROpennetMove(openentMove , myLastMove):
    cond1 = ((myLastMove==Rock)and(openentMove==Sic));
    cond2 = ((myLastMove==Paper)and(openentMove==Rock));
    cond3 = ((myLastMove==Sic)and(openentMove==Paper));
    
    return (cond1 or cond2 or cond3);
```

```python
def predictNextMove(openentMove , myLastMove):
    
    # if oppenet just coping move then send beating to it from dictionary
    if isCopyingMove(openentMove , myLastMove)==True:
        return beatingDic[myLastMove];
    
    # if oppenet just trying to send move which can beat your last move just reverse it.
    elif IsOpententRMove(openentMove , myLastMove)==True:
        return beatingDic[beatingDic[myLastMove]];
    
   # if oppenet is hoping you do step 2

    elif IsROpennetMove(openentMove , myLastMove)==True:
        return myLastMove;
            
    return 0;
```

```python
def opponentmove_agent(observation, configuration):
    global myLastMovH;
    if observation.step > 0:
        newMove= predictNextMove(observation.lastOpponentAction, myLastMovH)
        myLastMovH = newMove;
        return newMove;
    else:
        # most of player just retrun 0 for frist so try different :D
        myLastMovH = 1;
        return 1;
```
