This agent records the habits of the opponent's hand and puts out the best hand.

```python
%%writefile play_with_habit.py
import random

ROCK, PAPER, SCISSORS = 0, 1, 2
HAND = [ROCK, PAPER, SCISSORS]
DRAW, WIN, LOSE = 0, 1, 2

after_draw_opp_hand = [[0, 0, 0], [0, 0, 0], [0, 0, 0]]  # When the last is draw, the number of next hands (draw, paper, scissors)
after_win_opp_hand = [[0, 0, 0], [0, 0, 0], [0, 0, 0]]  # When the last is win, the number of next hands (draw, paper, scissors)
after_lose_opp_hand = [[0, 0, 0], [0, 0, 0], [0, 0, 0]]  # When the last is lose, the number of next hands (draw, paper, scissors)
last_judge, last_my_hand = None, None

def play_with_habit(observation, configuration):
    global after_draw_opp_hand
    global after_win_opp_hand
    global after_lose_opp_hand
    global last_judge
    global last_my_hand
    
    if last_my_hand:
        last_opp_hand = observation.lastOpponentAction
        # If the last time was a draw
        if (last_opp_hand - last_my_hand) % 3 == 0:
            # Get the most hands
            max_num = max(after_draw_opp_hand[last_opp_hand])
            max_hand = after_draw_opp_hand[last_opp_hand].index(max_num)
            # Record opp hand
            after_draw_opp_hand[last_opp_hand][last_opp_hand] += 1

        # If the last time was a win
        elif (last_opp_hand - last_my_hand) % 3 == 1:
            # Get the most hands
            max_num = max(after_win_opp_hand[last_opp_hand])
            max_hand = after_win_opp_hand[last_opp_hand].index(max_num)
            # Record opp hand
            after_win_opp_hand[last_opp_hand][last_opp_hand] += 1

        # If the last time was a lose
        else:
            # Get the most hands
            max_num = max(after_lose_opp_hand[last_opp_hand])
            max_hand = after_lose_opp_hand[last_opp_hand].index(max_num)
            # Record opp hand
            after_lose_opp_hand[last_opp_hand][last_opp_hand] += 1
            
        # next hand       
        my_hand = (max_hand + 1) % 3
        
    else:
        my_hand = random.choice(HAND)
        
    last_my_hand = my_hand

    return my_hand
```
