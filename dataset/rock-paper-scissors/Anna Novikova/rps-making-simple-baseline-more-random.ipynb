### About

I take the agent simply plays a hand that wins the last opponent's hand and mix it with completely random strategy.

```python
%%writefile submission.py

import random

def copy_opponent_agent (observation, configuration):
    # random choose of the strategy
    strategy = random.randint(0, 1) 
    if strategy == 0:        
        if observation.step > 0:
            opp_hand = observation.lastOpponentAction
            return (opp_hand + 1) % 3
        else:
            return random.randint(0, 2) 
    else:
        return random.randint(0, 2) 
```
