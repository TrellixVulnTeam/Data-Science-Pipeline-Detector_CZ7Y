## This method uses the pickle module, Python Standard Library.

```python
import pickle
```

## Generate an example of an object to be embedded

```python
import numpy as np


# fix random seed
np.random.seed(42)

# generate random 3x3 array(random seed fixed)
obj = np.random.randn(3, 3)
obj
```

## Convert numpy array to bytes

```python
# numpy array -> bytes object
obj_bytes = pickle.dumps(obj)
obj_bytes
```
