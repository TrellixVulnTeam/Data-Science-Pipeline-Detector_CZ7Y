## Explaination of the code-block below :

### An up vote will be highly appreciated. Thanks in advance

**Mapping**

0 : Rock

1 : Paper

2 : Scissors

your_action = (opponent_action+1)%3

**Simulate**

Opponent --------- You

0(rock)  --------- 1((0+1)%3 = 1 = paper)

1(paper) --------- 2((1+1)%3 = 2 = scissors)

2(scissors) ------ 3((2+1)%3 = 0 = rock)

So using this method you always tend to win. Your choice is based on opponents last action so tie/loss occurs when opponent chages move and in the next move you win.


```python
%%writefile submission.py

def copy_opponent_agent (observation, configuration):
    
    if observation.step > 0:
        opponent_action = observation.lastOpponentAction
        your_action = (opponent_action +1) % 3
        return your_action
    
    else:
        return 1
```
