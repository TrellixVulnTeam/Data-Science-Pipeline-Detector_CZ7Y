```python
%%writefile henny.py


from random import randrange
opp_history=[]
def henny(observation, configuration):
    global opp_history
    if observation['step']==0:
        opp_history = []
        return randrange(3)
    else:
        opp_history.append(observation["lastOpponentAction"])
        return((opp_history[randrange(len(opp_history))]+1)%3)
```
