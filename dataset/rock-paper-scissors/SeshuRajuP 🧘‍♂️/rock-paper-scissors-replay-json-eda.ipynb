```python
url = 'https://www.kaggleusercontent.com/episodes/4231325.json'
```

## Load Replay

```python
import urllib, json

response = urllib.request.urlopen(url)
data = json.loads(response.read())
```

## Explore Replay

```python
data.keys()
```

```python
data
```

```python
dict_obs = {}
for i in range(1000):
    player1 = data['steps'][i][0]
    player2 = data['steps'][i][1]
    dict_obs[i] = {'step':player1['observation']['step'],
                   'player1_action':player1['action'],'player1_reward':player1['reward'], 'player1_status':player1['status'],
                   'player2_action':player2['action'],'player2_reward':player2['reward'], 'player2_status':player2['status']
                  }
    if i != 0:
        dict_obs[i]['player1_lastOpponentAction'] = player1['observation']['lastOpponentAction']
        dict_obs[i]['player2_lastOpponentAction'] = player2['observation']['lastOpponentAction']
    else:
        dict_obs[i]['player1_lastOpponentAction'] = -1
        dict_obs[i]['player2_lastOpponentAction'] = -1
    
```

```python
df = pd.DataFrame(dict_obs)
df = df.T
df
```
