# Independent Random Agent Generator
Hypothesis: the nash equilibria for the rock, paper, scissors game is to choose each of rock, paper, & scissors with probability (1/3, 1/3, 1/3), therefore the optimal strategy is to use a random uniform agent.

Since all agents submitted to this competition are evaluated, this competition is more of a numbers game rather than a competition to find the best agent (assuming that random seeds cannot be easily found due to runtime containerization).

Therefore we would like to submit many random agents and make sure that each one produces different results.

## 1. Define the random agent

```python
%%writefile random_agent.py
import random
import time

# Arbitrary 'salt' to keep submissions far from time.time()
OFFSET = 99999

# Will be replaced based on the number of submissions made
submission_id = 1

random.seed(time.time() + OFFSET + submission_id)

def random_agent(obs, config):
    return random.randint(0,2)
```
