# Rock Paper Scissors - Random Agent

According to Game Theory, the Nash Equlibrium strategy for Rock Paper Scissors to always pick a random move, which makes for a very simple implemention.

```python
%%writefile submission.py

import random

def random_agent(observation, configuration):
    random.seed(None)
    return random.randint(0, configuration.signs-1)
```

```python
%run -i 'submission.py'
```

```python
from kaggle_environments import make

env = make("rps", configuration={"episodeSteps": 20}, debug=True)
# env.run(["submission.py", lambda obs, conf: random.randint(0, 2)])
env.run(["submission.py", random_agent])
print(env.render(mode="ansi", width=600, height=600))
```
