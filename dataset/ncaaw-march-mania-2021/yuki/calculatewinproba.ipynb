```python
def calculate_win_proba(df):
    df_new = df.copy(deep=True)
    df_new['WScore_mod'] = df_new['WScore'] // 2
    df_new['LScore_mod'] = df_new['LScore'] // 2
    df_new['AllScore_mod'] = df_new['WScore_mod'] + df_new['LScore_mod']
    df_new['WinScoreProbEstimate'] = df_new['WScore_mod']/(df_new['WScore_mod'] + df_new['LScore_mod'])
    df_new['WinProb'] = df_new.apply(lambda x: 1-binom.cdf(x['AllScore_mod']//2, x['AllScore_mod'], x['WinScoreProbEstimate']), axis=1)
    df_new.drop(['WScore_mod', 'LScore_mod', 'AllScore_mod', 'WinScoreProbEstimate'], inplace=True, axis=1)
    return df_new
```

```python
DATA_PATH = '../input/ncaaw-march-mania-2021/'
df_season_results = pd.read_csv(DATA_PATH + "WRegularSeasonCompactResults.csv")
```
