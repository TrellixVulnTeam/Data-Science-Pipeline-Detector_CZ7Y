```python
train = pd.read_csv("/kaggle/input/forest-cover-type-prediction/train.csv")
test = pd.read_csv("/kaggle/input/forest-cover-type-prediction/test.csv")
```

```python
Z = train.loc[:,'Elevation':'Horizontal_Distance_To_Fire_Points']

#converting range of values in continuous columns to between 0-1
for i in Z:
    maximum = np.max(train[i])
    train[i] = train[i]/maximum
    maximum_test = np.max(test[i])
    test[i] = test[i]/maximum_test
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
X = train.drop('Cover_Type',axis=1)
y = train['Cover_Type']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=40)
```

```python
for i in range(1,20):
    KNN = KNeighborsClassifier(n_neighbors = i)
    KNN.fit(X_train,y_train)
    print(KNN.score(X_test,y_test))
```

```python
KNN.fit(X_train,y_train)
```

```python
KNN.score(X_test,y_test)
```

```python
prediction = KNN.predict(test)
```

```python
submission = pd.DataFrame({'Id':test['Id'], 'Cover_Type':prediction})
submission.head()
```
