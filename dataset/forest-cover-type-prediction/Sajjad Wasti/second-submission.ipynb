```python
train_data =pd.read_csv('/kaggle/input/forest-cover-type-prediction/train.csv')
train_data.head()
```

```python
test_data = pd.read_csv('/kaggle/input/forest-cover-type-prediction/test.csv')
test_data.head()
```

```python
train_data.info()
```

```python
train_data['Cover_Type'].value_counts()
```

```python
X = train_data.drop(labels =['Id','Cover_Type'],axis=1)
Y = train_data['Cover_Type']
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_valid,y_train,y_valid = train_test_split(X,Y,random_state = 21)
```

```python
print(x_train.shape,x_valid.shape)
print(y_train.shape,y_valid.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(x_train,y_train)
```

```python
knn.score(x_valid,y_valid)
```

```python
new_test = test_data.drop(labels=['Id'],axis=1)
new_test.info()
```

```python
predict = knn.predict(new_test)
```

```python
Submission = pd.DataFrame(data = predict,columns=['Cover_Type'])
Submission.head()
```

```python
Submission['Id'] = test_data['Id']
Submission.set_index('Id',inplace = True)
```

```python
Submission.head()
```

```python
Submission.to_csv('Submission.csv')
```
