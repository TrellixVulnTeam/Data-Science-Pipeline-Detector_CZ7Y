```python
traindata = pd.read_csv('/kaggle/input/forest-cover-type-prediction/train.csv') 
testdata =pd.read_csv('/kaggle/input/forest-cover-type-prediction/test.csv')
```

```python
from sklearn.model_selection import train_test_split
X = traindata.drop(['Id', 'Cover_Type'], axis =1)
y = traindata['Cover_Type']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=25)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors= 5)
knn.fit(X_train, y_train)
```

```python
knn.score(X_test, y_test)
```

```python
pred = knn.predict(testdata.drop('Id', axis =1))
```

```python
submission = pd.DataFrame(data= pred, columns = ['Cover_Type'])
submission.head()
```

```python
submission['Id']= testdata['Id']
submission.set_index('Id',inplace=True)
```
