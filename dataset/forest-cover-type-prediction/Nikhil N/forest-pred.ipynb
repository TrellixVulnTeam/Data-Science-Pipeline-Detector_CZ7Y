```python
from sklearn.model_selection import train_test_split
```

```python
X = train_data.drop(labels=['Id','Cover_Type'], axis=1)
Y = train_data['Cover_Type']
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, random_state=42)
```

```python
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators= 70)
rfc.fit(x_train,y_train)
print(rfc.score(x_test,y_test)*100)
```

```python
predictions = rfc.predict(test_data.drop(labels=['Id'], axis=1))
```

```python
Submission = pd.DataFrame(data=predictions, columns=['Cover_Type'])
```

```python
Submission.head()
```

```python
Submission['Id'] = test_data['Id']
```
