## Dataset Information

* Target : predict the forest cover type .

The data is in raw form (not scaled) and contains binary columns of data for qualitative independent variables such as wilderness areas and soil type.

**Dataset field**
* Elevation (meters above mean sea level) - Elevation in meters

* Aspect (A slope which falls down to a deep valley on its western side and a shallower one on its eastern side has a westerly)- Aspect in degrees azimuth

* Slope - Slope in degrees

* Horizontal_Distance_To_Hydrology - Horz Dist to nearest surface water features
* Vertical_Distance_To_Hydrology - Vert Dist to nearest surface water features
* Horizontal_Distance_To_Roadways - Horz Dist to nearest roadway
* Hillshade_9am (0 to 255 index) - Hillshade index at 9am, summer solstice
* Hillshade_Noon (0 to 255 index) - Hillshade index at noon, summer solstice
* Hillshade_3pm (0 to 255 index) - Hillshade index at 3pm, summer solstice
* Horizontal_Distance_To_Fire_Points - Horz Dist to nearest wildfire ignition points
* Wilderness_Area (4 binary columns, 0 = absence or 1 = presence) - Wilderness area designation
* Soil_Type (40 binary columns, 0 = absence or 1 = presence) - Soil Type designation
* Cover_Type (7 types, integers 1 to 7) - Forest Cover Type designation


**The wilderness areas are:**

1. Rawah Wilderness Area
2. Neota Wilderness Area
3. Comanche Peak Wilderness Area
4.  Cache la Poudre Wilderness Area

**The seven types are:**

1. Spruce/Fir
2. Lodgepole Pine
3. Ponderosa Pine
4. Cottonwood/Willow
5. Aspen
6. Douglas-fir
7. Krummholz

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
 
model = RandomForestClassifier(n_estimators=120, random_state=42)

model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
y_predict = model.predict(test)
```

```python
final_prediction = pd.DataFrame()
final_prediction['Id'] = test['Id']
final_prediction['Cover_Type'] = y_predict
final_prediction.to_csv('submission.csv', index=False)
```
