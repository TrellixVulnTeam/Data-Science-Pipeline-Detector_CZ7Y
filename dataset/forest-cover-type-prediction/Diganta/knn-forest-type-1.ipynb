```python
train_df=pd.read_csv('/kaggle/input/forest-cover-type-prediction/train.csv')
train_df.head()
```

```python
test_df=pd.read_csv('/kaggle/input/forest-cover-type-prediction/test.csv')
test_df.head()
```

```python
test_id=test_df['Id']
```

```python
train_df.drop(['Id'],axis=1,inplace=True)
test_df.drop(['Id'],axis=1,inplace=True)
```

```python
train_df.shape
```

```python
test_df.shape
```

```python
X=train_df.drop(['Cover_Type'],axis=1)
y=train_df['Cover_Type']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=40)
```

```python
print(X_train.shape,y_train.shape)
print(X_test.shape,y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNN=KNeighborsClassifier(n_neighbors=6)
KNN.fit(X_train,y_train)
```

```python
KNN.score(X_test,y_test)
```

```python
pred=KNN.predict(test_df)
```

```python
result=pd.DataFrame(data=pred,columns=['Cover_Type'])
```

```python
result.head()
```

```python
result['Id']=test_id
result.set_index('Id',inplace=True)
result.head()
```

```python
result.to_csv('Submission.csv')
```
