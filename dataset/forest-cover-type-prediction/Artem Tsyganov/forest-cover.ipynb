```python
train_data = pd.read_csv("../input/forest-cover-type-prediction/train.csv")
test_data = pd.read_csv("../input/forest-cover-type-prediction/train.csv")
```

```python
train_data.head()
```

```python
X=train_data.drop(labels=['Id','Cover_Type'],axis=1)
y=train_data['Cover_Type']
```

**Train Test Split**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,  y_test = train_test_split(X, y, test_size = 0.2, random_state=40)
```

**Sklearn RidgeClassifier**

```python
from sklearn.linear_model import RidgeClassifier
from sklearn.metrics import accuracy_score

rc = RidgeClassifier()
rc.fit(X_train, y_train)
pred = rc.predict(X_test)
print("accuracy :", rc.score(X_test, y_test))
```

**Sklearn KNeighborsClassifier**

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
pred = knn.predict(X_test)
print("accuracy :", knn.score(X_test, y_test))
```
