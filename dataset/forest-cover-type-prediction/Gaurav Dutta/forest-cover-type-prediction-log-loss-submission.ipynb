```python
import pandas as pd
train=pd.read_csv("../input/forest-cover-type-prediction/train.csv")
train.head()
```

```python
train=train.drop(['Id'],axis=1)
train.head()
```

```python
Wilderness_Area_cols = [col for col in train.columns if 'Wilderness_Area' in col]
Wilderness_Area_cols
```

```python
Wilderness_Area=train[Wilderness_Area_cols]
Wilderness_Area.head()
```

```python
df_n = Wilderness_Area.apply(lambda x: x.idxmax(), axis = 1)
df_n.head()
```

```python
Wilderness_Area=pd.DataFrame(df_n, columns=['Wilderness'])
Wilderness_Area.head()
```

```python
Wilderness_Area.Wilderness.value_counts()
```

```python
Wilderness_Area.Wilderness=Wilderness_Area.Wilderness.map({'Wilderness_Area1':'Area1',
                                                           'Wilderness_Area2':'Area2',
                                                          'Wilderness_Area3':'Area3',
                                                          'Wilderness_Area4':'Area4'})
Wilderness_Area.Wilderness.value_counts()
```

```python
train=train.drop(['Wilderness_Area1','Wilderness_Area2','Wilderness_Area3','Wilderness_Area4'],axis=1)
train.head()
```

```python
train=pd.concat([train, Wilderness_Area], axis=1)
del Wilderness_Area,df_n
train.head()
```

```python
Soil_Type_cols = [col for col in train.columns if 'Soil_Type' in col]
Soil_Type_cols
```

```python
Soil_Type=train[Soil_Type_cols]
Soil_Type.head()
```

```python
df_n = Soil_Type.apply(lambda x: x.idxmax(), axis = 1)
df_n.head()
```

```python
Soil_Type=pd.DataFrame(df_n, columns=['Soil'])
Soil_Type.head()
```

```python
Soil_Type.Soil = Soil_Type.Soil.str.replace('Soil_', '')
Soil_Type.Soil.value_counts()
```

```python
train=train.drop(Soil_Type_cols,axis=1)
train.head()
```

```python
train=pd.concat([train, Soil_Type], axis=1)
del df_n,Soil_Type
train.head()
```

```python
train['Cover_Type'].value_counts()
```

```python
y_train = train.pop('Cover_Type')
X_train = train
```

```python
X_train.info()
```

```python
y_train.value_counts()
```

```python
from flaml import AutoML
automl = AutoML()
```

```python
automl.fit(X_train, y_train, task="classification",metric='log_loss',time_budget=3600)
```

```python
print('Best ML leaner:', automl.best_estimator)
print('Best hyperparmeter config:', automl.best_config)
print('Best log_loss on validation data: {0:.4g}'.format(automl.best_loss))
print('Training duration of best run: {0:.4g} s'.format(automl.best_config_train_time))
```
