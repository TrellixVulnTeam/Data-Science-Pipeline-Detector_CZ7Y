```python
test_data = pd.read_csv('/kaggle/input/forest-cover-type-prediction/test.csv')
```

```python
train_data = pd.read_csv('/kaggle/input/forest-cover-type-prediction/train.csv')
```

```python
train_data.head()
```

```python
print(test_data.shape)
print(train_data.shape)
```

```python
test_id = test_data['Id']
test_data.drop(['Id'] ,axis=1, inplace=True)
train_data.drop(['Id'],axis=1, inplace=True)
```

```python
train_data.info()
```

```python
train_data.isnull().sum()
```

```python
X = train_data.drop(['Cover_Type'], axis=1)
y = train_data['Cover_Type']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y)
```

```python
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(x_train , y_train)
```

```python
print(knn.score(x_test,y_test) * 100)
```

```python
pred = knn.predict(test_data)
```

```python
Sub = pd.DataFrame(data=pred,columns=['Cover_type'])
Sub.head()
```

```python
Sub['id'] = test_id
Sub.set_index('id',inplace=True)
```

```python
Sub.head()
```

```python
Sub.to_csv('Submission.csv')
```
