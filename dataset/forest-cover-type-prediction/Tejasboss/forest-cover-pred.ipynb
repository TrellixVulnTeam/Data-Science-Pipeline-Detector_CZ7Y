```python
train=pd.read_csv('/kaggle/input/forest-cover-type-prediction/train.csv')
test=pd.read_csv('/kaggle/input/forest-cover-type-prediction/test.csv')
train.info()
```

```python
non_categorical=train.drop(['Id','Cover_Type'],axis=1).columns[:15]
print(non_categorical)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
for col in non_categorical:
    train[col]=scaler.fit_transform(train[col].to_frame())
    test[col]=scaler.transform(test[col].to_frame())
```

```python
train.head(5)
```

```python
test.head(5)
```

```python
from sklearn.model_selection import train_test_split
X=train.drop(['Id','Cover_Type'],axis=1)
y=train['Cover_Type']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNN=KNeighborsClassifier(n_neighbors=11)
KNN.fit(X_train,y_train)
```

```python
KNN.score(X_test,y_test)
```

```python
predict=KNN.predict(test.drop('Id',axis=1))
```

```python
Submission=pd.DataFrame(data=predict,columns=['Cover_Type'])
Submission.head()
```

```python
Submission['Id']=test['Id']
Submission.set_index('Id',inplace=True)
```

```python
Submission.to_csv('Submission.csv')
```
