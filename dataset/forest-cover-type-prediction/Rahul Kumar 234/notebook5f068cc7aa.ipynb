```python
train_df=pd.read_csv("/kaggle/input/forest-cover-type-prediction/train.csv")
test_df=pd.read_csv("/kaggle/input/forest-cover-type-prediction/test.csv")
```

```python
column=train_df.columns[1:15]
```

```python
column
```

```python
for i in column:
    x=train_df[i].max()
    train_df[i]/=x
```

```python
train_df.head()
```

```python
X=train_df.drop(["Id","Cover_Type"],axis=1)
y=train_df["Cover_Type"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(1)
```

```python
knn.fit(X_train,y_train)
```

```python
knn.score(X_test,y_test)
```

```python
y_pred=knn.predict(test_df.drop('Id',axis=1))
```

```python
submission=pd.DataFrame(data=y_pred,columns=["Cover_Type"])
```
