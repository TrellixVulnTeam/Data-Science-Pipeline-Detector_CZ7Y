```python
traindata=pd.read_csv('/kaggle/input/forest-cover-type-prediction/train.csv')
traindata.head()
```

```python
testdata=pd.read_csv('/kaggle/input/forest-cover-type-prediction/test.csv')
testdata.head()
```

```python
print(traindata.shape)
print(testdata.shape)
```

```python
from sklearn.model_selection import train_test_split
X=traindata.drop(['Id','Cover_Type'],axis=1)
Y=traindata['Cover_Type']
Xtrain,Xtest,ytrain,ytest=train_test_split(X,Y)
```

```python
print(Xtrain.shape,Xtest.shape)
print(ytrain.shape,ytest.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=2)
knn.fit(Xtrain,ytrain)
```

```python
knn.score(Xtest,ytest)
```

```python
predict=knn.predict(testdata.drop('Id',axis=1))
```

```python
Submission=pd.DataFrame(data=predict,columns=['Cover_Type'])
Submission.head()
```

```python
Submission['Id']=testdata['Id']
Submission.set_index('Id',inplace=True)
```
