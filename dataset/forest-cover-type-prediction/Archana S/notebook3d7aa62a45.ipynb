```python
#saving this particular column to satisfy submission requirement
tst_id = tst_data['Id']
tst_id.shape 
```

```python
print(tst_data.shape)
print(tr_data.shape)
```

```python
tr_data.drop(['Id'], axis=1, inplace=True)
tst_data.drop(['Id'], axis=1, inplace=True)
print(tr_data.shape, tst_data.shape)
```

```python
X=tr_data.drop(['Cover_Type'],axis=1)
y=tr_data['Cover_Type']
  
```

```python
print(X.shape, y.shape)
```

```python
#splitting the train_data for training and evaluation data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y) 
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
print(knn.score(X_test, y_test))
pred = knn.predict(tst_data)

```

```python


print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
df=pd.DataFrame(data=pred,columns=['Cover_Type'])
df.head()
```

```python


df['Id']=tst_id
df.set_index('Id',inplace=True)
df.head()
```

```python

df.to_csv('df.csv')
```
