```python
train_data =  pd.read_csv("../input/forest-cover-type-prediction/train.csv")
train_data.head()
```

```python
test_data =  pd.read_csv("../input/forest-cover-type-prediction/test.csv")
test_data.head()
```

```python
train_data['Slope'].plot(kind = "hist")
```

```python
test_data['Slope'].plot(kind = "hist")
```

```python
x=train_data.drop(["Id","Cover_Type"],axis=1)
y=train_data['Cover_Type']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y ,random_state=40)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators = 70)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)*100
```

```python
predict=model.predict(test_data.drop(labels=['Id'],axis=1))
```

```python
submission= pd.DataFrame(data = predict,columns = ['Cover_Type'])
submission.head()
```

```python
submission['Id'] =test_data["Id"]
submission.set_index('Id',inplace=True)
```
