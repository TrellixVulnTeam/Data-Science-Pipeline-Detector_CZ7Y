```python
from sklearn.model_selection import train_test_split
X=train_data.drop(['Id','Cover_Type'],axis=1)
y=train_data['Cover_Type']
X_train,X_test,y_train,y_test=train_test_split(X,y)
```

```python
print(X_train.shape,y_train.shape)
print(X_test.shape,y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
for i in range(1,15):
    KNN=KNeighborsClassifier(n_neighbors=i)
    KNN.fit(X_train,y_train)
    print(f'Neighbour {i} : {KNN.score(X_test,y_test) * 100}')
```

```python
knn=KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,y_train)
knn.score(X_test,y_test)
```

```python
predict=knn.predict(test_data.drop('Id',axis=1))
```

```python
train_data['Cover_Type'].nunique()
```

```python
Submission=pd.DataFrame(data=predict,columns=['Cover_Type'])
Submission.head()
```

```python
Submission['Cover_Type'].nunique()
```

```python
Submission['Id']=test_data['Id']
Submission.set_index('Id',inplace=True)
```
