```python
import pandas as pd
import itertools
from spacy import displacy
all_perms = [''.join(list([str(x) for x in p])) for p in list(itertools.permutations(range(1,8), 7))]
LETTERS = {
    1: 'üéÖ',  # father christmas
    2: 'ü§∂',  # mother christmas
    3: 'ü¶å',  # reindeer
    4: 'üßù',  # elf
    5: 'üéÑ',  # christmas tree
    6: 'üéÅ',  # gift
    7: 'üéÄ',  # ribbon
    8: 'üåü',  # star
}
INV_LETTERS = {v: k for k, v in LETTERS.items()}
solution = pd.read_csv(sub)
strings = [''.join([str(INV_LETTERS[c]) for c in s]) for s in solution.schedule]
solution = pd.read_csv(wsub)
wstrings = [''.join([str(INV_LETTERS[c]) for c in s]) for s in solution.schedule]

def find_strings_perms_counts(strings, strings_perms, verbose=False):
    perms = []
    counts = [[],[],[]]
    perms_repate = [[],[],[]]
    perms_repate_counts = [[],[],[]]
    for s_i, s in enumerate(strings):
        perms.append([])
        counts[s_i] = [0]*len(s)
        for i in range(len(s)-6):
            p = s[i:i+7]
            if p in perms[s_i]:
                perms_repate[s_i].append(p)
            if p in strings_perms[s_i]:
                perms[s_i].append(p)
                for j in range(i, i+7):
                    counts[s_i][j] += 1
    for s_i, s in enumerate(strings):
        perms_repate_counts[s_i] = [0]*len(s)
        for i in range(len(s)-6):
            p = s[i:i+7]
            if p in perms_repate[s_i]:
                for j in range(i, i+7):
                    perms_repate_counts[s_i][j] += 1
            
    return perms, counts, perms_repate, perms_repate_counts
```

```python
group_perms, group_counts, group_perms_repate, perms_repate_counts = find_strings_perms_counts(strings, [all_perms,all_perms,all_perms])
wgroup_perms, wgroup_counts, wgroup_perms_repate, wperms_repate_counts = find_strings_perms_counts(wstrings, group_perms)
skip_perms = [list(set(group_perms[i])-set(wgroup_perms[i])) for i in range(3)]
sgroup_perms, sgroup_counts, sgroup_perms_repate, sperms_repate_counts = find_strings_perms_counts(strings, skip_perms)
```

```python
from spacy import displacy
def visualize_superpermutation(title, perm, counts):
    colors = {
            '0': '#ffffff',
            '1': '#f6f7da',
            '2': '#faf7da',
            '3': '#f0f7da',
            '4': '#c9df8a',
            '5': '#77ab59',
            '6': '#36802d',
            '7': '#234d20',
            'W': '#DC143C',
         }
    ents = []
    for i, count in enumerate(counts):
        if perm[i] == '8':
            count = 'W'
        ents.append({'start':i, 'end':i+1, 'label': str(count)})
        doc = {
            "text": perm,
            "ents": ents,
            "title": title
        }
    options = {"ents": [str(x) for x in ['0','1','2','3','4','5','6','7','W']], "colors": colors}
    displacy.render(doc, style="ent", options=options, manual=True, jupyter=True)
```
