```python
from functools import reduce
import pandas as pd
import numpy as np

df = pd.read_csv('../input/santa-2021/permutations.csv')

condition = df['Permutation'].str[:2] == 'ðŸŽ…ðŸ¤¶'
df_kick_off = df[condition]
df_permutations = df[~condition]

#Every line must contain all kick offs:
kick_offs = reduce(lambda x,y: x + y, df_kick_off.values, "")

#Split the other permutations.
permutations_split = np.array_split(df_permutations.values, 3)

# Solution = Kick offs + permutations
data = map(lambda x1: kick_offs + reduce(lambda x,y: x + y, x1, ""), permutations_split)

submission = pd.DataFrame(data, columns=['schedule'])
```
