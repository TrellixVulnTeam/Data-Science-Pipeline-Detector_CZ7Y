# This a quit easy understanding but efficient way to get a 2428 solution.

1. Split words
2. Rebalence to find out Exchaning Pair
3. Do Exchaning
4. Check permutations
5. -->2428 solution

**If you are quite good at coding, maybe you can make it an end-end way.**

```python
import itertools
import numpy as np
import pandas as pd
symbols = "üéÖü§∂ü¶åüßùüéÑüéÅüéÄ"
schedule = pd.read_csv("../input/best2440/submission-10.csv").schedule.tolist()
words = [s.translate(str.maketrans(symbols, "1234567")) for s in schedule]
words_f = words.copy()
list(map(len, words))
```

**Split all permutations begin with '12' (visualizing)**(This cell is a slightly update version of **SeshuRajuP**'s notebook for analyzing.)

```python
block = [[],[],[]]
prev = 0
for j in range(3):
    for i in range(len(words[j])-6):
        sub = words[j][i:i+7]
        if  ''.join(sorted(sub)) == '1234567' and sub[0:2] == '12':
            if i != 0:
                block[j].append(words[j][prev:i])
            prev = i
block[j].append(words[j][prev:i+7])
for i,e in enumerate(block):
    print(f'block[{i}]:\n{e}')
```

# Find out 3 Exchaning Pair from above
1. **block[0]<->block[1]**
2. **block[1]<->block[2]**
3. **block[0]<->block[2]**
**What is Exchaning Pair?**
for example:
* **'...|1523746', '1237456|...' in block[0] with '...|1623745', '1237465|...' in block[1]** (must in deferent blocks)is Called an **Exchaning Pair**, also, you can call it anything you want.
* When we doing **|1523746', '1237456|-->|1üåü237456|, block[0] lose '1523746' and add '1623745'** for certain. Same, doing **|1623745', '1237465|-->|1üåü237465|, block[1] lose '1623745' and add '1523746'**.block[0].add->block[1].lose,block[1].add->block[0].lose
* Finally, we shorten the total distance for 12 by doing exchaning like this for one time, In most case,except the **lose** and **add**, this will just lose duplicate permutations,So we won't lose any permutation actually*(But sometimes one block or more will lose one and the Exchaning Pair may won't show up directly, then we need to rebalance the block by regrouping strings were split above to get the lost permutation back, this can be fixed easily or toughly, I'll show you guys how to do this in the last part)*. If we do this 3 times, we shorten the total distance for 36, every block benefit 12.
* So, all we need to is to find out 3 Exchaning Pair. Actually there are more than just one Exchaning Pair included in every block of a 2440 solution, so if we find one is hard to get, then we can pass it and turn to another.
* Then, do exchaning, as below:

```python
for i in range(len(words[0])-13):
    if words[1][i:i+7]=='1325467':
        words[1] = words[1].replace(words[1][i:i+14],'1*254637')
    if words[0][i:i+7]=='1523746':
        words[0] = words[0].replace(words[0][i:i+14],'1*237456')
    if words[1][i:i+7]=='1623745':
        words[1] = words[1].replace(words[1][i:i+14],'1*237465')
    if words[0][i:i+7]=='1627543':
        words[0] = words[0].replace(words[0][i:i+14],'1*275463')
    if words[2][i:i+7]=='1327546':
        words[2] = words[2].replace(words[2][i:i+14],'1*275436')
    if words[2][i:i+7]=='1725463':
        words[2] = words[2].replace(words[2][i:i+14],'1*254673')
print('After exchaning:',[len(x) for x in words])
        
```

**Check out if we had lost any permutation after doing exchaning.**
**if didn't: count1=6    count2=36**

```python
mandatory = ['12'+''.join(x) for x in itertools.permutations(['3','4','5','6','7'], 5)]
num1=0
for p1 in mandatory:
    if p1 not in words[0] or p1 not in words[1] or p1 not in words[2]:
        num1+=1
        print(p1)
print('count1:',num1)

all_permutations = [''.join(x) for x in itertools.permutations(['1','2','3','4','5','6','7'], 7)]
num2=0
for p2 in all_permutations:
    if p2 not in words[0] and p2 not in words[1] and p2 not in words[2]:
        num2+=1
        print(p2)
print('count2:',num2)
```

# Example:
To get half part of '2376' Exchaning Pair in block[0]
* find the ...2376.' in block[0], find the split point, regroup:
* ...'12376542137654231765423716542376154237651423765'...'1237645'... -->
...'12376542137654231765423716542376154237651423765','1237645'...,...
* after regrouping block[0], we lose '5412376', then try to regroup block[1] or block[2] to get the lost one back,here we can regroup block[1]:

```python
mandatory = [''.join(x) for x in itertools.permutations(['1','2','3','4','5','6','7'], 7)]
dicf = {} #count every permutation
a1,b1,c1=words_f #2440 solution
for s in mandatory:
    dicf[s] = 0
    for i in range(len(a1)-6):
        if a1[i:i+7]==s:
            dicf[s]+=1
    for j in range(len(b1)-6):
        if b1[j:j+7]==s:
            dicf[s]+=1
    for k in range(len(c1)-6):
        if c1[k:k+7]==s:
            dicf[s]+=1
print('*******1*******lose:')
##split point1: 423765','123754 (from block[0])
x = '423765123754'
for i in range(len(x)-6):
    xt = x[i:i+7]
    if ''.join(sorted(xt))=='1234567':
        if dicf[xt]==1:
            print(xt)
print('*******2*******lose:')
##split point2: 273654','123764 (from block[0])
x = '273654123764'
for i in range(len(x)-6):
    xt = x[i:i+7]
    if ''.join(sorted(xt))=='1234567':
        if dicf[xt]==1:
            print(xt)
print('*******3*******lose:')
##split point3: 273654','126374(from blcok[1])
x = '2736541263745'
for i in range(len(x)-6):
    xt = x[i:i+7]
    if ''.join(sorted(xt))=='1234567':
        if dicf[xt]==1:
            print(xt)
##split point4: 263745', '123764(from block[1])
print('*******4*******lose:')
x = '263745123764'
for i in range(len(x)-6):
    xt = x[i:i+7]
    if ''.join(sorted(xt))=='1234567':
        if dicf[xt]==1:
            print(xt)
##when it's done, we lose '1423765' and add '1523764'


print('********match********') 
##find out where is the other part of '2376' Exchaning Pair. 
for i in range(len(a1)-6):
    if a1[i:i+7]=='1523764':
        print('block[0]',i,a1[i:i+7])
    if b1[i:i+7]=='1523764':
        print('block[1]',i,b1[i:i+7])
    if c1[i:i+7]=='1523764':
        print('block[2]',i,c1[i:i+7])
```
