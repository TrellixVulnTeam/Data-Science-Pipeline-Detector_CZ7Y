```python
def pattern_right(pattern,line,dels,xx,yy):
 #print(dels)
 pattern[f'con5{yy}_idx'] = str()
 pattern[f'con5{yy}_flg'] = str()
 pattern[f'con5{yy}_add'] = str()
 a=str()
 idx1 =0
 for i in pattern[f'con5{xx}_add']:
     if len(i)<= dels:
        i = 'x'
     i = i[-dels:]
     idx2=0
     for x in pattern[f'con5{xx}_add']:
         if len(x)<= dels:
            x= 'y'
         #print(x) 
         x =x[:dels]
         #if idx2 <5:
         #    print('#',x,i,len(x),len(i),dels,idx1,idx2)
         if x ==i:
             if a.find(i) == -1  and pattern[f'con5{yy}_flg'][idx1] =='' and idx1!=idx2:
                 #if idx1 ==0: #DEBUG
                 #   print(x,i,len(x),len(i),dels,idx1,idx2)
                 if  pattern[f'con5{yy}_flg'][idx2] =='' and pattern[f'con5{yy}_flg'][idx1]=='':
                     pattern[f'con5{yy}_flg'][idx2] =str('1')
                     pattern[f'con5{yy}_flg'][idx1] =str('-1') #block
                     pattern[f'con5{yy}_idx'][idx1]+= str(idx2)+' '
                     pattern[f'con5{yy}_add'][idx1]+=str(pattern[f'con5{xx}_add'][idx1])+str(pattern[f'con5{xx}_add'][idx2][-int(len(pattern[f'con5{xx}_add'][idx2]))+dels:])
                     a+= str(i)+' '
                     break
                
         idx2+=1
     idx1+=1
     #if idx1==2:
     #       stop

 
 a = pattern[pattern[f'con5{yy}_add']=='']

 pattern[f'con5{yy}_add'].iloc[a.index] = a[f'con5{xx}_add']

 aa = pattern.index[pattern[f'con5{yy}_flg']=='1'].to_list()
 #print(aa)
 #   display('droped')
 pattern = pattern.drop(pattern.index[aa]).reset_index(drop=True)
 #else:
 #   display(aa)

 a = str()
 for i in range(len(pattern)):
     a+=str(pattern[f'con5{yy}_add'].values[i])
 print('length=',len(a),'  step=',line,len(pattern),' ',dels)

 return pattern,len(a)
```
