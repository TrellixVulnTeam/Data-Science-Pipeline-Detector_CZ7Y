# Particle Swarm Optimization


In computational science, particle swarm optimization (PSO) is a computational method that optimizes a problem by iteratively trying to improve a candidate solution with regard to a given measure of quality. It solves a problem by having a population of candidate solutions, here dubbed particles, and moving these particles around in the search-space according to simple mathematical formula over the particle's position and velocity. Each particle's movement is influenced by its local best known position, but is also guided toward the best known positions in the search-space, which are updated as better positions are found by other particles. This is expected to move the swarm toward the best solutions.

Here is the link of the research paper : [Particle swarm optimization](https://ieeexplore.ieee.org/abstract/document/488968) by J. Kennedy and R. Eberhart

There are many variations of PSO. I will be using the simplest form of PSO.

# Let us define the PSO algorithm

### The main steps of the PSO algorithm for optimization are:
1. Initialization of the initial swarm population
2. Assign their inital positions in the search space randomly.
3. Assign their initial velocities randomly
4. Calculate the `local best` (the best solution for that individual in the swarm)
5. Calculate the `global best` (the best solution of the entire swarm)
6. Update the position for each step (`final_pos = inital_pos + vel`)
7. Update the velocity using the formula (used in the program)
8. Update the `local best` and the `global best`
9. Repeat from step 6 until iterations end or objective is fulfilled.

```python
class PSO:
    def __init__(self, iterations=10, swarm_size=10, w=1, c1=2.05, c2=2.05, chi=0.729, UB=None, LB=None, logging_steps=10, obj_fun=None):
        self.iterations = iterations
        self.swarm_size = swarm_size
        self.w = w
        self.c1 = c1
        self.c2 = c2
        self.chi = chi
        self.best_params = None
        self.logging_steps = logging_steps
        self.fun = obj_fun
        self.UB = UB
        self.LB = LB
        
        if self.fun is None:
            raise AttributeError('Must define a objective function to optimize')

    
    def optimize(self):
        
        def Individual(particle):
            f=self.fun(particle)
            particle['fx']=f
            return particle
        
        def random_particle():
            pos=[]
            velocity=[]

#             i = random.randint(1, 7)
#             j = random.randint(1, 7)
#             while(i!=j):
#                 j = random.randint(1, 7)

#             t1 = random.randint(2000, 2700)
#             t2 = random.randint(t1+1, 5000)

#             adj1 = random.randint(0,20)
#             adj2 = random.randint(0,20)

#             pos = [i, j, t1, t2, adj1, adj2]

            for i in range(len(self.UB)):
                pos.append(random.randint(self.LB[i], self.UB[i]))
                velocity.append(random.randint(0, 10))

            return {'pos':pos, 'vel':velocity}
        
        def update(particle):
            vel=particle['vel']
            pos=particle['pos']
            pbest=particle['pbest']
            gbest=particle['gbest']
            #position update

            new_pos=[]
            for i in range(6):
                op = [1, -1]
                temp=pos[i]+vel[i]* (op[random.randint(0,1)])
                if temp>self.UB[i]:
                    new_pos.append(self.LB[i])
        #             new_pos.append(random.randint(LB[i], UB[i]))
                elif temp<self.LB[i]:
                    new_pos.append(self.LB[i])
        #             new_pos.append(random.randint(LB[i], UB[i]))
                else:
                    new_pos.append(temp)

            #vel update
            new_vel=[]
            for i in range(len(vel)):
                new_vel.append( int(self.chi * ((self.w * vel[i]) + self.c1 * np.random.random() * (pbest[i]-pos[i]) + \
                               self.c2 * np.random.random() * (gbest[i]-pos[i]))))


            return new_pos, new_vel
    

        
        
        SWARM=[Individual(random_particle()) for i in range(self.swarm_size)]

        gbest=sorted(SWARM, key=lambda x: x['fx'])[0]['pos']

        for particle in SWARM:
            particle['gbest']=gbest
            particle['pbest']=particle['pos']

        F=[]
        I=[]
        
        for i in range(self.iterations):
            for particle in SWARM:

                old_fx=particle['fx']

                new_pos, new_vel=update(particle)
                updated_fx = self.fun({'pos':new_pos})
                particle['pos']=new_pos
                particle['vel']=new_vel

                if old_fx<updated_fx:
                    particle['pbest']=new_pos

                particle['fx']=updated_fx


            new_gbest=sorted(SWARM, key=lambda x: x['fx'])[0]['pos']
            new_gbest_fx=self.fun({'pos':new_gbest})

            old_gbest_fx=self.fun({'pos':SWARM[0]['gbest']})

            if new_gbest_fx<old_gbest_fx:
                for particle in SWARM:
                    particle['gbest']=new_gbest

            if i%self.logging_steps == 0:
                best_score = self.fun({'pos':SWARM[0]['gbest']})
                self.best_params = SWARM[0]['gbest']
                print(f'Iteration: {i} | Best params: {self.best_params} | Max len : {best_score}')

            I.append(i+1)
            F.append(self.fun({'pos':particle['gbest']})) 
        
        print('\n'*2)
        fig = go.Figure(go.Scatter(x=I, y=F, mode='lines'))
        fig.show()

        
    def best_params(self):
        return self.best_params
```

# Define Objective function (must)

We need to pass a objective function that return the value we want to optimize.

Let use now define the objective function.

Here the objective function is returning the max length of the 3 strings. There are several parameters like changing the symbol values of )()(  (i and j), selecting good split points (t1 and t2) and selecting the overlap length (adj1 and adj2)

```python
permutations = [''.join(x) for x in itertools.permutations(['1','2','3','4','5','6','7'], 7)]


def fun(particle):
    pos = particle['pos']
    i, j, t1, t2, adj1, adj2 = pos
    
    if j==i:
        return 10000

    if t1>t2:
        return 10000
    
    strings = [solution[:t1+adj1], solution[t1-adj1:t2+adj2], solution[t2-adj2:]]
    all_strings = ['1', '2', '3','4','5','6','7']
    all_strings.remove(str(i))
    all_strings.remove(str(j))
    required_permutaions = [str(i)+str(j) + ''.join(x) for x in itertools.permutations(all_strings, 5)]

    for i in range(3):
        for permutation in required_permutaions:
            if permutation not in strings[i]:
                strings[i]+=permutation
    all_lens = []


    for p in permutations:
        if p not in strings[0] and p not in strings[1] and p not in strings[2]:
            return 10000

    for s in strings:
        all_lens.append(len(s))

    return max(all_lens)
```
