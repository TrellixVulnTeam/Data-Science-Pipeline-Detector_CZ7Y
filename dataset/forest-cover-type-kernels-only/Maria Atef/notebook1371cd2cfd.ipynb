```python
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

from sklearn.linear_model import Perceptron
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble  import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
df = pd.read_csv('../input/forest-cover-type-kernels-only/train.csv.zip',
                 encoding='utf-8') 
# Preview the first 5 lines of the loaded data 

df.head()
```

```python
df.drop('Id',
  axis='columns', inplace=True)
```

```python
df.insert(0,"mm",df["Horizontal_Distance_To_Roadways"]+df["Horizontal_Distance_To_Hydrology"]+df["Horizontal_Distance_To_Hydrology"])
```

```python

y=df.values[: , -1] -1
X = df.values[: , 0:-1]
```

```python
df.isnull().sum()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler

stdsc = StandardScaler()
X_train_std = stdsc.fit_transform(X_train)
X_test_std = stdsc.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier(criterion='gini',
                                n_estimators=325, 
                                random_state=1,max_depth=34)
forest.fit(X_train_std, y_train)
accuracy_score(forest.predict(X_test_std), y_test)
```

```python


















# Preprocess Test Data
df_submit = pd.read_csv('../input/forest-cover-type-kernels-only/test.csv.zip')
id_submit = df_submit.values[:, 0]
df_submit = df_submit.drop(columns='Id')

## put your feature engineering here but for df_submit ##
#-------------------------------------------------------#
#########################################################
df_submit.insert(0,"mm",df_submit["Horizontal_Distance_To_Roadways"]+df_submit["Horizontal_Distance_To_Hydrology"]+df_submit["Horizontal_Distance_To_Hydrology"])


X_submit = df_submit.values
X_submit_std = stdsc.transform(X_submit)


print("Evaluation..")
y_submit = forest.predict(X_submit_std)

submission = {'Id':id_submit, 'Cover_Type': (y_submit + 1).astype(int)}
df_submission = pd.DataFrame(submission)
df_submission.to_csv('submission.csv', index=False)


```
