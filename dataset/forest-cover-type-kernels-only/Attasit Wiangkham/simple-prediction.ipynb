```python
#import library
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import  LogisticRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.neural_network import MLPClassifier
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
```

import time for monitoring training time

```python

import time
```

Import data

```python
# Import data
data = pd.read_csv('../input/for-train/ML-Train.csv',index_col="Id")
```

Selected input (x) and target (y)
Split data for train and test process (70%:30%)

```python
x = data.drop('Cover_Type',axis=1)
y = data['Cover_Type']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3,random_state=42)
```

Standardization data using sklearn preprocessing

```python
#Standardization
scaler = StandardScaler()
STDTrain = scaler.fit_transform(x_train)
STDTest = scaler.fit_transform(x_test)
```

Start time count for training and testing model process

```python
start_time = time.time()
```

The model included 3 algorithms to predict clover type
    1. Decision tress
    2. LogisticRegression with polynomial feature
    3. Neural network models (backpropagation)

```python
#Training
#Decision tree
Tre = DecisionTreeClassifier()
Tre = Tre.fit(STDTrain, y_train)
```

```python
#Training
#LogisticRegression
poly = PolynomialFeatures(2)
STDTrain_2 = poly.fit_transform(STDTrain)
STDTest_2 = poly.fit_transform(STDTest)
Poly = LogisticRegression()
Poly = Poly.fit(STDTrain_2,y_train)
```

```python
#Training
#Neural network models
NN = MLPClassifier(solver='lbfgs', alpha=5e-3,
                   hidden_layer_sizes=(10, 10, 10), random_state=42)
NN.fit(STDTrain,y_train)
```

```python
#Test
#Decision tree
TPre = Tre.predict(STDTest)
#LogisticRegression
LPre = Poly.predict(STDTest_2)
#Neural network models
NPre = NN.predict(STDTest)
```

Stop time count for train and test process

```python
finish_time = time.time()
time = finish_time-start_time
```

Measure accuracy of model using metrics.accuracy_score

```python
#Accuracy
print ('Modelling process')
print('Accuracy decision tress = ', metrics.accuracy_score(y_test,TPre))
print('Accuracy logisticRegression = ', metrics.accuracy_score(y_test,LPre))
print('Accuracy neural network models = ', metrics.accuracy_score(y_test,NPre))
print('Training time', time)
```
