```python
import os
import pandas as pd

df = pd.read_csv('../input/forest-cover-type-kernels-only/train.csv.zip', header=None, encoding='utf-8')
df.tail()
```

```python
df_forest = pd.read_csv('../input/forest-cover-type-kernels-only/train.csv.zip', )

df_forest
```

```python
df_forest=df_forest.drop(columns='Id')
df_forest
```

```python

X = df_forest.values[:, 0:-1]
y = df_forest.values[:,-1] - 1
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
from sklearn.svm import SVC

svm = SVC(kernel='rbf', random_state=1, gamma=0.000001, C=10000)
svm.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(svm.predict(X_test), y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier(criterion='gini',
                                n_estimators=10000, 
                                random_state=1,
                                n_jobs=2)
forest.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(forest.predict(X_test), y_test)
```
