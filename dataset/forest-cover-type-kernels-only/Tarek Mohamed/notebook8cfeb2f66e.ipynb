```python
import os
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
```

```python
import pandas as pd
df = pd.read_csv('../input/forest-cover-type-kernels-only/train.csv.zip')
df.drop('Id',axis=1, inplace=True)
df.insert(3,'roed+fire',df['Horizontal_Distance_To_Roadways']+df['Horizontal_Distance_To_Fire_Points'])# confirm
df.insert(3,'Ele-dist_to_hyd',df['Elevation']-df['Vertical_Distance_To_Hydrology'])
df.insert(3,'test',df['Vertical_Distance_To_Hydrology']/(df['Horizontal_Distance_To_Hydrology']+1))
df.head()
```

```python
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values -1
```

```python
sc=StandardScaler().fit(X)
X_std=sc.transform(X)
```

```python
forest=RandomForestClassifier(n_estimators=351,max_depth=27,random_state=1).fit(X=X_std,y=y)
```

```python

# Preprocess Test Data
df_submit = pd.read_csv('../input/forest-cover-type-kernels-only/test.csv.zip')
id_submit = df_submit.values[:, 0]
df_submit = df_submit.drop(columns='Id')

df_submit.insert(3,'roed+fire',df_submit['Horizontal_Distance_To_Roadways']+df_submit['Horizontal_Distance_To_Fire_Points'])# confirm
df_submit.insert(3,'Ele-dist_to_hyd',df_submit['Elevation']-df_submit['Vertical_Distance_To_Hydrology'])
df_submit.insert(3,'test',df_submit['Vertical_Distance_To_Hydrology']/(df_submit['Horizontal_Distance_To_Hydrology']+1))


X_submit = df_submit.values
sc=StandardScaler().fit(X_submit)
X_submit_std = sc.transform(X_submit)


print("Evaluation..")
y_submit = forest.predict(X_submit_std)

submission = {'Id':id_submit, 'Cover_Type': (y_submit + 1).astype(int)}
df_submission = pd.DataFrame(submission)
df_submission.to_csv('submission.csv', index=False)


```
