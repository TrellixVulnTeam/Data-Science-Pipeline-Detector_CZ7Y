# Fonction pour changer les chiffres du Cover_Type en leur nom respectif

```python
def forest(x):
    if x==1:
        return 'Spruce/Fir'
    elif x==2:
        return 'Lodgepole Pine'
    elif x==3:
        return 'Ponderosa Pine'
    elif x==4:
        return 'Cottonwood/Willow'
    elif x==5:
        return 'Aspen'
    elif x==6:
        return 'Douglas-fir'
    elif x==7:
        return 'Krummholz'
    
train['Cover_Type'] = train['Cover_Type'].apply(lambda x: forest(x))
```
