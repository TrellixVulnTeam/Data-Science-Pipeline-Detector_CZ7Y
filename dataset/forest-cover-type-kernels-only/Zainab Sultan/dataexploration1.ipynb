We check for outliers. Outliers are points which are far from the norm of the data. These can confuse our classifier. They can result from noise in measurememt, or just be a valid data point since uniformity is rare in real life contexts. However they shouldnt be ignored.

```python

z = np.abs(stats.zscore(train))

print(len(np.unique(np.where(z >= 3)[0])) )
```

There are 13155 data points that are outliers. Existence of outliers might confuse the model. There are multiple ways to deal with outliers such as imputing them or removing them.
