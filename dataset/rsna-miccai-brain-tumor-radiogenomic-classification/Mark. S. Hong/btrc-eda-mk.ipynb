# Get number of files in the train set

```python
for i in id_list:
    tmp_path = PATH+'/train/'+i
    sub_list = os.listdir(tmp_path) #"FLAIR", "T1w", "T1wCE", "T2w"
    sub_dict = {}
    sum = 0
    for j in sub_list:
        tmp_tmp_path = tmp_path+"/"+j
        len_list = len(os.listdir(tmp_tmp_path))
        sub_dict[j] = len_list
        sum += len_list
        sub_dict["sum"] = sum
    dict_list[str(i)] = sub_dict
```

# Get orientation of images grouped by sequences

## Correct the deviation of ImageOrientationPatient tags

```python
for i in id_list:
    tmp_path = PATH+"/train/"+i
    sub_list = os.listdir(tmp_path) #"FLAIR", "T1w", "T1wCE", "T2w"
    sub_dict_position = {}
    for j in sub_list:
        rt = "unknown"
        tmp_tmp_path = tmp_path+"/"+j
        (x1,y1,_,x2,y2,_) = pydicom.dcmread(tmp_tmp_path+'/'+os.listdir(tmp_tmp_path)[0]).ImageOrientationPatient
        
    
        if (x1<1 and x1>0.7) and (y1<0.7 and y1>-0.7) and (x2<0.7 and x2>-0.7) and (y2<0.7 and y2>-0.7):
            rt = 'coronal'
        if (x1<1 and x1>0.7) and (y1<0.7 and y1>-0.7) and (x2<0.7 and x2>-0.7) and (y2<1 and y2>0.7): # (1,0,0,1):
            rt = 'axial'
        if (x1<0.7 and x1>-0.7) and (y1<1 and y1>0.7) and (x2<0.7 and x2>-0.7) and (y2<0.7 and y2>-0.7):
            rt = 'sagittal'
        
#         if (x1,y1,x2,y2) == (1,0,0,0):# 
#             rt = 'coronal'
#         if (x1,y1,x2,y2) == (1,0,0,1):
#             rt = 'axial'
#         if (x1,y1,x2,y2) == (0,1,0,0):
#             rt = 'sagittal'
        sub_dict_position[j] = rt
    dict_list2[str(i)] = sub_dict_position
```
