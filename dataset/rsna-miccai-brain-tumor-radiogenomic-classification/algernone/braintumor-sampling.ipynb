```python
SEED = 42

pl.utilities.seed.seed_everything(SEED, workers=True)
```

```python
df = pd.read_csv("../input/rsna-miccai-brain-tumor-radiogenomic-classification/train_labels.csv", dtype={'BraTS21ID': str})
print(df.shape)
df = df[~df.BraTS21ID.isin(['00109', '00123', '00709'])].reset_index()
print(df.shape)
```

```python
df.head(3)
```

```python

skf = StratifiedKFold(random_state=SEED, shuffle=True)

df['fold'] = -1
for fold, (train_i, val_i) in enumerate(skf.split(df, df.MGMT_value)):
    df.loc[val_i, 'fold'] = fold
    
df.fold.value_counts()
```
