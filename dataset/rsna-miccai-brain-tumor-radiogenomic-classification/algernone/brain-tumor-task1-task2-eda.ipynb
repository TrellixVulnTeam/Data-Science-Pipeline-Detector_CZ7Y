```python
shapes = set()
for file in tqdm(glob('./nifti/train/*')):
    for nii in glob(f'{file}/*'):
        nb_file = nb.load(nii)
        shapes.add(nb_file.shape)
print(shapes)
```

```python
task1_study = [file for file in task1_files if '00000' in file][0]
task2_study = [file for file in task2_files if '00000' in file][0]

```

```python
task1_study_files = glob(f'{task1_study}/*')
files_n = len(task1_study_files)

fig = plt.figure(figsize=(18, 18))
for i, file in enumerate(task1_study_files, 1):
    nb_file = nb.load(file)
    print(file, nb_file.shape)
    ax = fig.add_subplot(1, files_n, i)
    ax.imshow(nb_file.get_fdata()[:,:,77])
```

```python
task1_study_files = glob(f'{task2_study}/*')
files_n = len(task1_study_files)

fig = plt.figure(figsize=(18, 18))
for i, file in enumerate(task1_study_files, 1):
    nb_file = nb.load(file)
    print(file, nb_file.shape)
    ax = fig.add_subplot(1, files_n, i)
    ax.imshow(nb_file.get_fdata()[:,:,77])
```
