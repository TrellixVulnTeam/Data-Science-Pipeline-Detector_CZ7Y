```python
import pandas as pd
from sklearn.model_selection import StratifiedKFold
```

```python
df = pd.read_csv('../input/rsna-miccai-brain-tumor-radiogenomic-classification/train_labels.csv')
# Exclusion of three cases from training cohort
# https://www.kaggle.com/c/rsna-miccai-brain-tumor-radiogenomic-classification/discussion/262046
df = df[~df['BraTS21ID'].isin([109, 123, 709])]
df = df.sort_values(by='BraTS21ID')
df = df.reset_index(drop=True)
```

```python
cv = StratifiedKFold(n_splits=3, random_state=42, shuffle=True)
X, y = df['BraTS21ID'].values, df['MGMT_value'].values

for i, (train_idx, val_idx) in enumerate(cv.split(X, y)):
    df.loc[val_idx, 'fold'] = i
df['fold'] = df['fold'].astype(int)
```

```python
df.to_csv('split.csv', index=False)
df.head(20)
```
