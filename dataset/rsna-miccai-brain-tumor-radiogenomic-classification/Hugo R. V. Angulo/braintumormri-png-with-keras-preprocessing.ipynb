# ****Uploading PNG files****

Upload png dataset from:
https://www.kaggle.com/jonathanbesomi/rsna-miccai-png

Use the keras.preprocessing module from:
https://www.tensorflow.org/tutorials/load_data/images#load_using_tfkeraspreprocessing

and assigned it to training data:

```python
import pathlib
dr='../input/rsna-miccai-png/train'
train_ds = tf.keras.preprocessing.image_dataset_from_directory(
  dr,
  labels='inferred',
  validation_split=0.2,
  subset="training",
  seed=123,
  # image_size=(img_height, img_width),
  # batch_size=batch_size,
)
```

So, we can see that each class corresponds to patient's id codes. 

*If someone knows how could I change the clases to the competition labels (MGMT_value), please leave it on comments...*

```python
class_names = train_ds.class_names
print(class_names)
```

# Patient's ID label

So, we can visualize the images and the patient id:

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 10))
for images, labels in train_ds.take(1):
  for i in range(9):
    ax = plt.subplot(3, 3, i + 1)
    plt.imshow(images[i].numpy().astype("uint8"))
    plt.title(class_names[labels[i]])
    plt.axis("off")
```

We can choose another directory to show the images and the type of images as the label if we set a patient's id in the file's path

```python
dr2 ='../input/rsna-miccai-png/train/00000'
img_type = tf.keras.preprocessing.image_dataset_from_directory(
  dr2,
  labels='inferred',
  validation_split=0.2,
  subset="training",
  seed=123,
  # image_size=(img_height, img_width),
  # batch_size=batch_size,
)
```

```python
class_names2 = img_type.class_names
print(class_names2)
```

So, we can visualize several types of images of the 00000 patient:

```python
plt.figure(figsize=(10, 10))
for images, labels in img_type.take(1):
  for i in range(12):
    ax = plt.subplot(4, 4, i + 1)
    plt.imshow(images[i].numpy().astype("uint8"))
    plt.title(class_names2[labels[i]])
    plt.axis("off")
```
