```python
import numpy as np 
import pandas as pd 
import tensorflow as tf 
import matplotlib.pyplot as plt 
import cv2 as cv
from path import Path
import os 
import glob
import tensorflow_hub as hub
import os 
```

```python
train_df= pd.read_csv('../input/rsna-miccai-brain-tumor-radiogenomic-classification/train_labels.csv')
sample_df = pd.read_csv('../input/rsna-miccai-brain-tumor-radiogenomic-classification/sample_submission.csv')
```

```python
train_df.head()
```

```python
train_df['BraTS21ID'][-2:]
```

```python
files = glob.glob('../input/rsna-miccai-png/train/*/FLAIR/*.png')
```

```python
from PIL import Image
```

```python
# def get_images(x):
#     print(x)
#     image = tf.io.read_file(x)
#     image = tf.io.decode_image(image,expand_animations = False)
# #     image = np.array(Image.open(x).convert('RGB'))\
#     image = tf.keras.preprocessing.image.img_to_array(image)


#     return image 

def get_images(x):
    print(x)
    image = tf.io.read_file(x)
    image = tf.io.decode_image(image,expand_animations = False,channels = 1)
    image = tf.image.resize(image , [128,128])
    return image 
```

```python
plt.imshow(get_images(files[0]))
get_images(files[0]).shape
```

```python
train_inputs = tf.data.Dataset.from_tensor_slices(files).map(get_images)
```

```python
Path(files[0]).parts()[4]
```

```python
mgmt_values = []

for x in files:
    brat_id = int(Path(x).parts()[4])
    mgmt_values.append(train_df[train_df['BraTS21ID']==brat_id]['MGMT_value'].to_numpy()[0])
    
```

```python
train_outputs = tf.data.Dataset.from_tensor_slices(mgmt_values)
```

```python
train_outputs.element_spec
```

```python
train_dataset = tf.data.Dataset.zip((train_inputs,train_outputs)).batch(20).prefetch(tf.data.AUTOTUNE)
```

```python
# base_layer = hub.KerasLayer('https://tfhub.dev/google/efficientnet/b5/classification/1',trainable = False)
```

```python

inputs = tf.keras.layers.Input(shape = (128,128,1))
# x = tf.keras.layers.Lambda(lambda x : cv.merge((x,x,x)))(inputs)
# x = base_layer(inputs)
x = tf.keras.layers.Conv2D(128,3,activation ='relu')(inputs)
x = tf.keras.layers.GlobalMaxPooling2D()(x)
x = tf.keras.layers.Flatten()(x)
x = tf.keras.layers.Dense(128, activation ='relu')(x)
x = tf.keras.layers.Dense(64,activation = 'relu')(x)
x = tf.keras.layers.Dense(32, activation = 'relu')(x)

# input_2 = tf.keras.layers.Input(shape = [3,])
# print(input_2.shape)
# y = tf.keras.layers.Flatten(input_2)
# y = tf.keras.layers.Dense(128,activation ='relu')(input_2)

# concat = tf.keras.layers.concatenate([x,y])

# z = tf.keras.layers.Dense(64,activation= 'relu')(concat)

outputs = tf.keras.layers.Dense(1, activation = 'softmax')(x)

model =tf.keras.Model(inputs,outputs)
model.summary()
```

```python
model.compile(loss = tf.keras.losses.binary_crossentropy,
             optimizer = tf.keras.optimizers.Adam(),
             metrics = ['accuracy'])
```

```python
model.fit(train_dataset,
         epochs = 15)
```

```python
test_files = glob.glob('../input/rsna-miccai-png/test/*/FLAIR/*.png')
```

```python
test_dataset = tf.data.Dataset.from_tensor_slices(test_files).map(get_images).batch(32)
```

```python
predictions = model.predict(test_dataset)
```
