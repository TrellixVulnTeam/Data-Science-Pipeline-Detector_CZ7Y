```python
import pandas as pd
from fastai.vision.all import *

path = Path('../input/rsna-miccai-brain-tumor-radiogenomic-classification')
```

```python
train_df = pd.read_csv('../input/rsna-miccai-brain-tumor-radiogenomic-classification/train_labels.csv')
train_df
```

```python
# Exclusion of three cases from training cohort
# https://www.kaggle.com/c/rsna-miccai-brain-tumor-radiogenomic-classification/discussion/262046
# 00109
# 00123
# 00709

#show dropped 
display(train_df[train_df['BraTS21ID'].isin([109, 123, 709])])
train_df = train_df[~train_df['BraTS21ID'].isin([109, 123, 709])]
```

```python
temp = []
temp1 = []
for i in train_df.iterrows():
    imageid = str(i[1][0]).zfill(5)
    x = path/'train'/imageid
    for j in os.listdir(x):
        for k in os.listdir(x/j):
            temp.append([imageid, j, k, str(i[1][1]), x/j/k])

train_dcm_df = pd.DataFrame(temp,columns=['BraTS21ID', "Method", "FileName", 'MGMT_value', 'filepath'])
train_dcm_df.to_csv('train_dcm.csv', index=False)
train_dcm_df
```

```python
png_path = Path('../input/rsna-miccai-png')

temp = []
for i in train_df.iterrows():
    imageid = str(i[1][0]).zfill(5)
    x = png_path/'train'/imageid
    for j in os.listdir(x):
        num = len(os.listdir(x/j))
        if num >= 20:
            images = sorted([int(pngFile.split('.')[0].split('-')[1]) for pngFile in os.listdir(x/j)])[num-10:num+10]
        else:
            images = sorted([int(pngFile.split('.')[0].split('-')[1]) for pngFile in os.listdir(x/j)])
        for image_num in images:
            pngPath = x/j/('Image-' + str(image_num) + '.png')
            temp1.append([imageid, j, image_num, str(i[1][1]), pngPath])
        for k in os.listdir(x/j):
            temp.append([imageid, j, k, str(i[1][1]), x/j/k])

train_png_df = pd.DataFrame(temp,columns=['BraTS21ID', "Method", "FileName", 'MGMT_value', 'filepath'])
train_png_df.to_csv('train_png.csv', index=False)
train_png_df
```

```python
train_midPng_df = pd.DataFrame(temp1,columns=['BraTS21ID', "Method", "ImageNum", 'MGMT_value', 'filepath'])
train_midPng_df.to_csv('train_mid_png.csv', index=False)
train_midPng_df
```
