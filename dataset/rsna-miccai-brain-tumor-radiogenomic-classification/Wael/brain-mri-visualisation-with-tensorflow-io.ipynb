```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
import tensorflow_io as tfio
```

```python
import glob
```

we test the viz with the patient with the id 0000

```python
data_directory = '../input/rsna-miccai-brain-tumor-radiogenomic-classification'
scan_id = "00000"
split = "train"
mri_type = 'T2w'
```

we store the list of slices 

```python
slices = sorted(glob.glob(f"{data_directory}/{split}/{scan_id}/{mri_type}/*.dcm"))
```

```python
lossy_image = []
for i in slices : 
    image_bytes_i = tf.io.read_file(i)
    image_bytes_i.shape
    skipped_i = tfio.image.decode_dicom_image(image_bytes_i, on_error='skip', dtype=tf.uint8)
    lossy_image_i = tfio.image.decode_dicom_image(image_bytes_i, scale='auto', on_error='lossy', dtype=tf.uint8)
    lossy_image.append(lossy_image_i)
print(lossy_image_i.shape)
```

```python
len(lossy_image)
```

we have here 408 slice in the MRI folder

## viz of the slice 150

```python
plt.imshow(np.squeeze(lossy_image[150].numpy()), cmap='gray')
```

## viz of the slice 230

```python
plt.imshow(np.squeeze(lossy_image[230].numpy()), cmap='gray')
```
