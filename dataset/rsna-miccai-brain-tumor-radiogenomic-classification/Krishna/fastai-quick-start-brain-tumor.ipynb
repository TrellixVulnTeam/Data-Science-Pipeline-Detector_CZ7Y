This notebook describe the steps to quick start with FastAI. The dataset is based on the fatastic work by Jonathan [here](https://www.kaggle.com/c/rsna-miccai-brain-tumor-radiogenomic-classification/discussion/253000) 

```python
from fastai.vision.all import *
```

```python
from fastai import test_utils
test_utils.show_install()
```

```python
def seed_everything(seed=0):
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.backends.cudnn.deterministic = True

seed_everything()
```

```python
path = Path('../input/rsna-miccai-png/'); path.ls()
```

```python
dls = ImageDataLoaders.from_folder(path, train='train',
                                   item_tfms=Resize(224),valid_pct=0.2,
                                   bs=64,seed=0)
```

```python
dls.show_batch()
```

```python
print(dls.vocab); print(dls.c)
```

```python
learn = cnn_learner(dls, resnet34, metrics=[error_rate, accuracy], model_dir="/tmp/model/").to_fp16()
```

```python
learn.lr_find()
```

```python
learn.fit_one_cycle(1, lr_max=1e-2, cbs=ShortEpochCallback())
```

```python
learn.show_results()
```
