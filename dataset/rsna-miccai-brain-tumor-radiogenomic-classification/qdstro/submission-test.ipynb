# Prerequisites

```python
import SimpleITK as sitk
import os
import pandas as pd
```

```python
# Create list of test patients
patients = sorted(os.listdir('../input/rsna-miccai-brain-tumor-radiogenomic-classification/test/'))
print(patients)
```

```python
# Function to load DICOM Series

reader = sitk.ImageSeriesReader()
reader.LoadPrivateTagsOn()

def load(path):
    
    filenamesDICOM = reader.GetGDCMSeriesFileNames(path)
    reader.SetFileNames(filenamesDICOM)
    loadedImage = reader.Execute()
    
    return loadedImage
```

# Preprocessing

```python
sequences = [('FLAIR', 0), ('T1w', 1), ('T2w', 3)]

for p in patients:
    
    try:
    
        print(f'Processing {p}')
        print('T1wCE')

        T1Norm = load(f'/kaggle/input/rsna-miccai-brain-tumor-radiogenomic-classification/test/{p}/T1wCE')
        sitk.WriteImage(T1Norm, f'/kaggle/working/{p}_RES_0002.nii.gz')

        for s, i in sequences:

            print(s)
            
            Img = load(f'/kaggle/input/rsna-miccai-brain-tumor-radiogenomic-classification/test/{p}/{s}')
            sitk.WriteImage(Img, f'/kaggle/working/{p}_RES_000{i}.nii.gz')
        
    except Exception as e:
        
        print(f'could not process {p}')
        print(e)
```

# Inference
