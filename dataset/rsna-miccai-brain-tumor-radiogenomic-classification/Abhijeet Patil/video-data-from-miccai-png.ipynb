```python
import numpy as np
import pandas as pd
import os
import cv2
import numpy as np
import glob
from tqdm import tqdm
```

```python
for dir_ in sorted(glob.glob('../input/rsna-miccai-png/*/*')):
    train, id_ = (dir_.split('/')[-2:])
    try: os.makedirs(f'dataset/{train}/{id_}')
    except: pass
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX

for dir_ in tqdm(glob.glob('../input/rsna-miccai-png/*/*/*')):
    train, id_, name_ = dir_.split('/')[-3:]
    save_vid = f'dataset/{train}/{id_}/{name_}.avi'

    img_array = []
    d = {int(image.split('/')[-1].split('.')[0].split('-')[-1]):image for image in sorted(glob.glob(f'{dir_}/*.png'))}
    for num, filename in sorted(d.items()):
        img = cv2.imread(filename)
        cv2.putText(img,f'{num}',(10,500), font, 1,(255,255,255),2)
        height, width, layers = img.shape
        size = (width,height)
        img_array.append(img)

    out = cv2.VideoWriter(save_vid, cv2.VideoWriter_fourcc(*'DIVX'), 5, size)
    for i in range(len(img_array)):
        out.write(img_array[i])
    out.release()
```
