# Determining the Number of Images for Each Modality Type

```python
train_path = ('../input/rsna-miccai-brain-tumor-radiogenomic-classification/train')

modes = ['FLAIR', 'T1w', 'T1wCE', 'T2w']
```

```python
image_counts = {m : [] for m in modes}

for i in tqdm(train.index):
    row = train.loc[i, :]
    folder = row['dir']
    
    for m in modes:
        temp = os.listdir(f'{train_path}/{folder}/{m}')
        print(temp)
        image_counts[m].append(len(temp))
        
for m in modes:
    train[m] = image_counts[m]

train.head()
```
