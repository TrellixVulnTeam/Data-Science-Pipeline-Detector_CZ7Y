```python
import glob
import os
import pydicom
import numpy as np
from pydicom import dcmread
import matplotlib.pyplot as plt
folder = "../input/rsna-miccai-brain-tumor-radiogenomic-classification/test/00161/"


paths = []

for i, t in enumerate(("FLAIR", "T1w", "T1wCE", "T2w"), 1):
    t_paths = sorted(glob.glob(os.path.join(folder, t, "*")))
    for file in t_paths:
        paths.append(file)
```

```python
for i in range(0,10):
    print(dcmread(paths[i]))
```

```python
def load_dcm(path):
    dicom = pydicom.read_file(path)
    data = dicom.pixel_array
    data = data - np.min(data)
    if np.max(data) != 0:
        data = data / np.max(data)
    data = (data * 255).astype(np.uint8)
    return data


def visualize_dcm():
    plt.figure(figsize=(16, 5))
    for i, t in enumerate(("FLAIR", "T1w", "T1wCE", "T2w"), 1):
        t_paths = sorted(glob.glob(os.path.join(folder, t, "*")))
        for file in t_paths:
            data = load_dcm(file)
            plt.subplot(1, 4, i)
            plt.imshow(data, cmap="gray")
            plt.title(f"{t}", fontsize=16)
            plt.axis("off")
    plt.show()
```

```python
visualize_dcm()
```
