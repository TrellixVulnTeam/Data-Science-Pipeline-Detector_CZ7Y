## Resampling nifti

```python
from nilearn import image
```

```python
directory = './DatasetNew/Train/Flair/'
resampled_flair='./DatasetNew/Train_resam/Flair'
for filename in os.listdir(directory):
   # print(filename,directory)
    f = os.path.join(directory, filename)
    sample_img = nib.load(f)
    affine=np.array([[-0.5 ,0 , 0 , 0] ,[0,0.5,0,0] , [0,0,3,0] , [0,0,0,1]])
    resample_img=nilearn.image.resample_img(sample_img, target_affine=affine, 
                                        target_shape=None, interpolation='continuous', 
                                        copy=True, order='F', clip=True, fill_value=0,
                                        force_resample=False)
    path = '{}'.format(filename)
    cp = os.path.join(resampled_flair, path)
    print(cp)
    nib.save(resample_img,cp)
```
