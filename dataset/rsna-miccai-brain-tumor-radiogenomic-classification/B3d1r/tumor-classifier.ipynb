```python
def decider_func(y_pred,exist_value = 1):
    if exist_value == 1:
        result = []
        farklar = []
        for i in y_pred:
            if i[0]> i[1]:
                result.append(1)
                fark = i[0]-i[1]
                farklar.append(fark)

            elif i[1] > i[0]:
                result.append(0)
                fark = i[1]-i[0]
                farklar.append(fark)

        result_tuple = (result.count(1),result.count(0))
        farklar_index = []
        for i in range(len(farklar)):
            if farklar[i] < 0.8:
                farklar_index.append(i)

        revised = []
        for i in range(len(result)):
            if i not in farklar_index:
                revised.append(result[i])

        revised_tuple =(revised.count(1),revised.count(0))
        if revised_tuple[0] > revised_tuple[1]:

            decision = 1

        else:
            decision = 0
        return decision,revised_tuple
    else:
        return -1
```

```python
#driver_code
predictions = []
for index in range(len(iter_liste)):
    
    y_pred_FLAIR = tester_function(index,model_FLAIR,"FLAIR")
    decision_FLAIR,revised_tuple_FLAIR = decider_func(y_pred_FLAIR)
    
    y_pred_T1w = tester_function(index,model_T1w,"T1w")
    decision_T1w,revised_tuple_T1w =decider_func(y_pred_T1w)
    
    y_pred_T1wCE = tester_function(index,model_T1wCE,"T1wCE")
    decision_T1wCE,revised_tuple_T1wCE = decider_func(y_pred_T1wCE)
    
    y_pred_T2w = tester_function(index,model_T2w,"T2w")
    decision_T2w,revised_tuple_T2w = decider_func(y_pred_T2w)
    
    liste_dec = [decision_FLAIR,decision_T1w,decision_T1wCE,decision_T2w]
    print(liste_dec)
    liste_calc = []
    liste_revised_tuple = [revised_tuple_FLAIR,revised_tuple_T1w,revised_tuple_T1wCE,revised_tuple_T2w]
    liste_weights = []
    for i in liste_revised_tuple:
        if i[0] > i[1]:
            weight = i[0] / (i[1]+1)
            
        else:
            weight = i[1] / (i[0]+1)
            
        liste_weights.append(weight)
    summa = 0
    for i in range(len(liste_dec)):
        
        if liste_dec[i] == 0:
            calc = liste_weights[i] * (-1)
            liste_calc.append(calc)
            summa += calc
            
        else:
            calc = liste_weights[i] * (1)
            liste_calc.append(calc)
            summa += calc
            
    
    if summa >0:
        decision = 1
        
    else:
        decision = 0
        
    
    predictions.append(decision)
    print(summa)
    print("done")
```
