## **Process:**

Run with multiple processes.
For each tar file:

    Download tar and md5
    Checksum
    Extract tar file
    Delete tar file
    Use OpenCV to read the image and save in new location as 448x448
    Delete original image.

After download and resize img, [move all images into one folder](https://www.kaggle.com/yaxi1031/merge-500-downloaded-train-folders-into-one-folder)

Zip that foder and copy it to google driver.
