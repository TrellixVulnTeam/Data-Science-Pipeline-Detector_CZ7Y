This exploration uses data from the [Austin Animal Shelter][1] and the purpose of this project is to predict outcomes for shelter animals.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

train_animals=pd.read_csv("../input/train.csv")
sample=pd.read_csv("../input/sample_submission.csv")
test_animals=pd.read_csv("../input/test.csv")
print(train_animals.head())
```

First lets clean the data up a bit a separate age into 4 columns (years, months, weeks, days)

```python
type_count=train_animals["AnimalType"].value_counts()
breed_count=train_animals["Breed"].value_counts()
color_count=train_animals["Color"].value_counts()
type_count.plot(kind="bar")
print(color_count)
```

First lets clean up the data a bit and create an additional columns.  We'll create a column to distinguish pure breeds from mutts. 

```python

pure_mutt=[]
breed=np.array(train_animals["Breed"])
for i in breed:
    if ("Mix" in i) | ('/' in i):
        pure_mutt.append("Mutt")
    else:
        pure_mutt.append("Pure")

train_animals["Pure/Mutt"]=pure_mutt        
print(train_animals[["Pure/Mutt", "Breed"]].head(200))
```

Have you heard of [Black Dog Syndrome][1]?
Shelters often refer to black dog syndrome, because darker colored dogs  tend to stay longer at the shelter. Lets create a column indicating if the dog was black to explore if darker dogs have different outcomes.
