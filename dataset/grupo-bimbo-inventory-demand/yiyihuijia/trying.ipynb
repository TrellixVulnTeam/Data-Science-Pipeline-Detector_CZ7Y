```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
column_Semana = pd.read_csv("../input/train.csv", usecols=['Semana'])
print(column_Semana.describe())
plt.hist(column_Semana['Semana'].tolist())
plt.title('Distribution of Semana')
plt.xlabel('Semana')
plt.ylabel('count')
plt.show()
```

```python
colum_target=pd.read_csv("../input/train.csv",usecols=['Demanda_uni_equil'])
print(colum_target.describe())
plt.hist(colum_target['Demanda_uni_equil'].tolist())
plt.title('Distribution of Target')
plt.xlabel('target')
plt.ylabel('count')
plt.show()
sns.distplot(colum_target['Demanda_uni_equil'].tolist())
```

```python
m=colum_target['Demanda_uni_equil'].mean()
result_mean=pd.read_csv("../input/test.csv",usecols=['id'])
result_mean['Demanda_uni_equil']=m
result_mean.to_csv('result_mean.csv',index=False)
```
