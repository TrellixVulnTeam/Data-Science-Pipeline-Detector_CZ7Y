```python
#Importar paquetes

import cv2
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import matplotlib.pyplot as plt 
%matplotlib inline


# Para ver nuestro directorio 

import os
import random
import gc #Para limpiar memoria

```

```python
train_dir = '../input/dogs-vs-cats-redux-kernels-edition/train.zip'
test_dir = '../input/dogs-vs-cats-redux-kernels-edition/test.zip'

train_dog = ['../input/dogs-vs-cats-redux-kernels-edition/train.zip/{}'.format(i) for i in os.listdir(train_dir) if 'dog' in i] #obtener imagenes de perro
train_cat = ['../input/dogs-vs-cats-redux-kernels-edition/train.zip/{}'.format(i) for i in os.listdir(train_dir) if 'cat' in i] #obtener imagenes de gato

test_img = ['../input/dogs-vs-cats-redux-kernels-edition/test.zip/{}'.format(i) for i in os.listdir(test_dir)] #obtener imagenes de prueba

train_img = train_dog[:2000] + train_cat[:2000] #llenar el dataset de imagenes de cada uno (2000)
random.shuffle(train_img) #barajear las imagenes de manera aleatoria 

#limpiar listas que ya no se usan 
del train_dog
del train_cat
gc.collect() #Colectar basura para guardar memoria 
```

```python
#Declarar las nuevas dimensiones de las imagenes para estandarizarlas
nrows = 150
ncolums = 150
channels = 3 
#Son 3 canales ya que usamos imagenes a color, porque lo que sabemos que se componen de RGB
```

```python
#Funcion para leer y procesar las imagenes 
def read_and_process_image(list_of_images):

    X = [] #Arreglo de imagenes
    Y = [] #Arreglo de etiquetas

    for image in list_of_images:
        X.append(cv2.resize(cv2.imread(image, cv2.IMREAD_COLOR), (nrows,ncolums), interpolation=cv2.INTER_CUBIC)) #leer la imagen
        #obtener etiquetas
        if 'dog' in image:
            Y.append(1)
        if 'cat' in image:
            Y.append(0)
        
        
return X,Y
```
