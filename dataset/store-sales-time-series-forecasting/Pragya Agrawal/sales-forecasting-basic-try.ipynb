```python
#import required packages
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# read all datasets
he_df = pd.read_csv('../input/store-sales-time-series-forecasting/holidays_events.csv')
oil_df = pd.read_csv('../input/store-sales-time-series-forecasting/oil.csv')
stores_df = pd.read_csv('../input/store-sales-time-series-forecasting/stores.csv')
trans_df = pd.read_csv('../input/store-sales-time-series-forecasting/transactions.csv')
train_df = pd.read_csv('../input/store-sales-time-series-forecasting/train.csv')
test_df = pd.read_csv('../input/store-sales-time-series-forecasting/test.csv')
sub_df = pd.read_csv('../input/store-sales-time-series-forecasting/sample_submission.csv')
```

```python
                                                ## TRAIN ##
# merge store dataset
train_df1 = pd.merge(train_df, stores_df, on = 'store_nbr')

# merging columns
train_df1["city_state"] = train_df1["city"] + '_' + train_df1["state"]

train_df1['cluster'] = train_df1['cluster'].astype(str)
train_df1["type_cluster"] = train_df1["type"] + '_' + train_df1["cluster"]

train_df1['store_nbr'] = train_df1['store_nbr'].astype(str)
train_df1["store_nbr_type"] = train_df1["store_nbr"] + '_' + train_df1["type"]

#train_df1.drop(['city', 'state', 'type', 'cluster'], axis = 1, inplace = True)  


                                                ## TEST ##
# merge store dataset
test_df1 = pd.merge(test_df, stores_df, on = 'store_nbr')

# merging columns
test_df1["city_state"] = test_df1["city"] + '_' + test_df1["state"]

test_df1['cluster'] = test_df1['cluster'].astype(str)
test_df1["type_cluster"] = test_df1["type"] + '_' + test_df1["cluster"]

test_df1['store_nbr'] = test_df1['store_nbr'].astype(str)
test_df1["store_nbr_type"] = test_df1["store_nbr"] + '_' + test_df1["type"]

#test_df1.drop(['city', 'state', 'type', 'cluster'], axis = 1, inplace = True)  
```

```python
del train_df
del test_df
```

```python
train_df1.rename({'type': 'store_type'}, axis=1, inplace=True)
test_df1.rename({'type': 'store_type'}, axis=1, inplace=True)
```

```python
                                                ## TRAIN ##
# merge holiday_events dataset
train_df1 = pd.merge(train_df1, he_df, on = 'date', how = 'left')

# merge columns
train_df1["holiday"] = train_df1["locale"] + '_' + train_df1["locale_name"] + '_' + train_df1["type"]
train_df1.rename({'type': 'he_type'}, axis=1, inplace=True)


                                                ## TEST ##
# merge holiday_events dataset
test_df1 = pd.merge(test_df1, he_df, on = 'date', how='left')

# merge columns
test_df1["holiday"] = test_df1["locale"] + '_' + test_df1["locale_name"] + '_' + test_df1["type"]
test_df1.rename({'type': 'he_type'}, axis=1, inplace=True)
```

```python
                                                ## TRAIN ##
# dropping extra column 'id'
train_df1.drop('id', axis = 1, inplace = True)

# setting date column as index
train_df1['date'] = pd.to_datetime(train_df1['date'], errors='coerce')
train_df1 = train_df1.set_index('date')

                                                ## TEST ##
sub = pd.DataFrame()
sub['id'] = test_df1['id']

# dropping extra column 'id'
test_df1.drop('id', axis = 1, inplace = True)

# setting date column as index
test_df1['date'] = pd.to_datetime(test_df1['date'], errors='coerce')
test_df1 = test_df1.set_index('date')
```

```python
print(train_df1.isna().any())
print(test_df1.isna().any())
print(train_df1.isnull().any())
print(test_df1.isnull().any())
```

```python
                                                ## TRAIN ##
train_df1.store_nbr = train_df1.store_nbr.astype(int)
train_df1.cluster = train_df1.cluster.astype(int)

                                                ## TEST ##
test_df1.store_nbr = test_df1.store_nbr.astype(int)
test_df1.cluster = test_df1.cluster.astype(int)
```

```python
                                                ## TRAIN ##
train_df1["family"] = train_df1["family"].astype('category')
train_df1["family"] = train_df1["family"].cat.codes

train_df1["city"] = train_df1["city"].astype('category')
train_df1["city"] = train_df1["city"].cat.codes

train_df1["state"] = train_df1["state"].astype('category')
train_df1["state"] = train_df1["state"].cat.codes

train_df1["store_type"] = train_df1["store_type"].astype('category')
train_df1["store_type"] = train_df1["store_type"].cat.codes

train_df1["city_state"] = train_df1["city_state"].astype('category')
train_df1["city_state"] = train_df1["city_state"].cat.codes

train_df1["type_cluster"] = train_df1["type_cluster"].astype('category')
train_df1["type_cluster"] = train_df1["type_cluster"].cat.codes

train_df1["store_nbr_type"] = train_df1["store_nbr_type"].astype('category')
train_df1["store_nbr_type"] = train_df1["store_nbr_type"].cat.codes

train_df1["he_type"] = train_df1["he_type"].astype('category')
train_df1["he_type"] = train_df1["he_type"].cat.codes

train_df1["locale"] = train_df1["locale"].astype('category')
train_df1["locale"] = train_df1["locale"].cat.codes

train_df1["locale_name"] = train_df1["locale_name"].astype('category')
train_df1["locale_name"] = train_df1["locale_name"].cat.codes

train_df1["description"] = train_df1["description"].astype('category')
train_df1["description"] = train_df1["description"].cat.codes

train_df1["holiday"] = train_df1["holiday"].astype('category')
train_df1["holiday"] = train_df1["holiday"].cat.codes

train_df1["transferred"] = train_df1["transferred"].astype('category')
train_df1["transferred"] = train_df1["transferred"].cat.codes

                                                ## TEST ##
test_df1["family"] = test_df1["family"].astype('category')
test_df1["family"] = test_df1["family"].cat.codes

test_df1["city"] = test_df1["city"].astype('category')
test_df1["city"] = test_df1["city"].cat.codes

test_df1["state"] = test_df1["state"].astype('category')
test_df1["state"] = test_df1["state"].cat.codes

test_df1["store_type"] = test_df1["store_type"].astype('category')
test_df1["store_type"] = test_df1["store_type"].cat.codes

test_df1["city_state"] = test_df1["city_state"].astype('category')
test_df1["city_state"] = test_df1["city_state"].cat.codes

test_df1["type_cluster"] = test_df1["type_cluster"].astype('category')
test_df1["type_cluster"] = test_df1["type_cluster"].cat.codes

test_df1["store_nbr_type"] = test_df1["store_nbr_type"].astype('category')
test_df1["store_nbr_type"] = test_df1["store_nbr_type"].cat.codes

test_df1["he_type"] = test_df1["he_type"].astype('category')
test_df1["he_type"] = test_df1["he_type"].cat.codes

test_df1["locale"] = test_df1["locale"].astype('category')
test_df1["locale"] = test_df1["locale"].cat.codes

test_df1["locale_name"] = test_df1["locale_name"].astype('category')
test_df1["locale_name"] = test_df1["locale_name"].cat.codes

test_df1["description"] = test_df1["description"].astype('category')
test_df1["description"] = test_df1["description"].cat.codes

test_df1["holiday"] = test_df1["holiday"].astype('category')
test_df1["holiday"] = test_df1["holiday"].cat.codes

test_df1["transferred"] = test_df1["transferred"].astype('category')
test_df1["transferred"] = test_df1["transferred"].cat.codes
```
