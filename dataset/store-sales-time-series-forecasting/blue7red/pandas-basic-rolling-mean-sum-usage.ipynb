```python
import numpy as np
import pandas as pd
```

```python
OIL_PATH = "../input/store-sales-time-series-forecasting/oil.csv"
```

```python
oil = pd.read_csv(OIL_PATH, parse_dates=['date'], infer_datetime_format=True, index_col='date')
```

```python
oil.head()
```

# fill mean data 

```python
 oil['dcoilwtico'] = oil['dcoilwtico'].fillna( oil['dcoilwtico'].mean())
```

# 1.rolling(1) 1 day mean and sum 

```python
oil['oil_1day_mean'] = oil['dcoilwtico'].rolling(1).mean()
```

```python
oil['oil_1day_sum'] = oil['dcoilwtico'].rolling(1).sum()
```

# 2.rolling(2) 2 days mean and sum 

```python
oil['oil_2day_mean'] = oil['dcoilwtico'].rolling(2).mean()
```

```python
oil['oil_2day_sum'] = oil['dcoilwtico'].rolling(2).sum()
```

# 3.rolling(3) 3 days mean and sum 

```python
oil['oil_3day_mean'] = oil['dcoilwtico'].rolling(3).mean()
```

```python
oil['oil_3day_sum'] = oil['dcoilwtico'].rolling(3).sum()
```

# 4.rolling(7) 7 days mean and sum 

```python
oil['oil_oneweek_mean'] = oil['dcoilwtico'].rolling(7).mean()
```

```python
oil['oil_oneweek_sum'] = oil['dcoilwtico'].rolling(7).sum()
```
