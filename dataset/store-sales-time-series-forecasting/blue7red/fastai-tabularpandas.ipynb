```python
from fastai.tabular.all import *
from sklearn.metrics import mean_squared_log_error
```

```python
path = Path('../input/store-sales-time-series-forecasting')
Path.BASE_PATH = path
path.ls()
```

```python
train_df = pd.read_csv(path/'train.csv')
transaction_df = pd.read_csv(path/'transactions.csv')
test_df = pd.read_csv(path/'test.csv')
sub_df = pd.read_csv(path/'sample_submission.csv')
oil_df = pd.read_csv("../input/store-sales-time-series-forecasting/oil.csv")
stores_df = pd.read_csv("../input/store-sales-time-series-forecasting/stores.csv")

train_df.shape, transaction_df.shape, test_df.shape, sub_df.shape
```

```python
train_df = train_df.merge(oil_df, on = 'date', how='left')
train_df = train_df.merge(stores_df, on = 'store_nbr', how='left')
# train_df = train_df.merge(df_trans, on = ['date', 'store_nbr'], how='left')

test_df = test_df.merge(oil_df, on = 'date', how='left')
test_df = test_df.merge(stores_df, on = 'store_nbr', how='left')
```

```python
splits = RandomSplitter(valid_pct=0.2)(range_of(train_df))
```

```python
tab_pd = TabularPandas(train_df, 
                       procs = [Categorify, FillMissing, Normalize],
                       cat_names=['store_nbr', 'family','city','type'],
                       cont_names =['onpromotion'],
                       y_names = 'sales',
                       splits = splits)
```

```python
dls = tab_pd.dataloaders(bs=512)
```

```python
len(dls.train), len(dls.valid)
```

```python
dls.show_batch()
```

```python
train_df.sales.max()
```

```python
def rmsle(inp, targ):
    inp = inp.detach().cpu().numpy()
    targ = targ.detach().cpu().numpy()
    return np.sqrt(mean_squared_log_error(targ, inp))
```

```python
learner = tabular_learner(dls, metrics=rmsle, y_range=(0, 125000))
```

```python
learner.fit_one_cycle(1)
```

```python
dl = learner.dls.test_dl(test_df)
```

```python
preds = learner.get_preds(dl=dl)
preds[0].shape
```

```python
sub_df.sales = preds[0].detach().cpu().numpy()
sub_df.head(3)
```

```python
sub_df.to_csv('submission.csv', index=False)
```
