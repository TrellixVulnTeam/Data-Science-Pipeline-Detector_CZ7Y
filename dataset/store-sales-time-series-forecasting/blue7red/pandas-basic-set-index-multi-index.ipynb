```python
import numpy as np
import pandas as pd
```

```python
TRAIN_PATH = "../input/store-sales-time-series-forecasting/train.csv"
```

# date : datetime64[ns] => by nanosecond

```python
train = pd.read_csv(TRAIN_PATH,
                       usecols=['store_nbr', 'family', 'date', 'sales'],
                       dtype={'store_nbr': 'category', 'family': 'category', 'sales': 'float32'},
                       parse_dates=['date'], infer_datetime_format=True)

print(train.info())
train.head()
```

# date : period[D] => by day
#### Y => year 
#### M => Month
#### W => Week
#### D => Day

### h => hour 
### m => minute
### s => second

```python
train.date = train.date.dt.to_period('D')
print(train.info())
train.head()
```

# MultiIndex 
### ['store_nbr', 'family', 'date']

```python
train = train.set_index(['store_nbr', 'family', 'date']).sort_index()
print(train.info())
train
```
