```python
import numpy as np
import pandas as pd
```

```python
OIL_PATH = "../input/store-sales-time-series-forecasting/oil.csv"
```

# 1. default

```python
oil = pd.read_csv(OIL_PATH)
print(oil.info())
oil.head()
```

# 2. using index_col

### index_col -> to index

```python
oil = pd.read_csv(OIL_PATH,index_col='date')
print(oil.info())
oil.head()
```

# 3. parse_dates + index_col

### parse_dates - > to datetime 
### index_col -> to index

```python
oil = pd.read_csv(OIL_PATH,parse_dates=['date'],index_col='date')
print(oil.info())
oil.head()
```

# 4.parse_dates + infer_datetime_format + index_col

### parse_dates - > to datetime 
### infer_datetime_format -> parsing speed is fast 
### index_col -> to index

infer_datetime_formatbool, ***default False***

If True and no format is given, attempt to infer the format of the datetime strings based on the first non-NaN element, 

### and if it can be inferred, switch to a faster method of parsing them. 
### **In some cases this can increase the parsing speed by ~5-10x.**

```python
oil = pd.read_csv(OIL_PATH, parse_dates=['date'], infer_datetime_format=True, index_col='date')
print(oil.info())
oil.head()
```

# set day of week

```python
oil["dayofweek"] = oil.index.dayofweek
oil.head(10)
```
