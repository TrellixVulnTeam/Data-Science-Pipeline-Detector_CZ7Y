```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
from dataprep.eda import plot, plot_correlation, create_report, plot_missing
from pandas_profiling import ProfileReport
```

<h1 style='background:#CCE2CB; border:0; color:black'><center> Read Data </center></h1> 

```python
df_holiday = pd.read_csv('../input/store-sales-time-series-forecasting/holidays_events.csv')
df_oil = pd.read_csv('../input/store-sales-time-series-forecasting/oil.csv')
df_store = pd.read_csv('../input/store-sales-time-series-forecasting/stores.csv')
df_transaction = pd.read_csv('../input/store-sales-time-series-forecasting/transactions.csv')
df_train = pd.read_csv('../input/store-sales-time-series-forecasting/train.csv')
df_test = pd.read_csv('../input/store-sales-time-series-forecasting/test.csv')
```

<h1 style='background:#CCE2CB; border:0; color:black'><center> MERGE DATA </center></h1> 

```python
train = df_train.merge(df_oil, on='date')
train = train.merge(df_holiday[['date', 'type', 'transferred']], on='date')
train = train.merge(df_store, on='store_nbr')
```

```python
train.rename(columns={'type_x':'holiday_type', 'type_y':'store_type', 'dcoilwtico':'oil_price'}, inplace=True)
```

```python
train.head()
```

```python
train['date'] = pd.to_datetime(train['date'])
```

```python
train['Year'] = train.date.dt.year
train['Year-Month'] = train['date'].apply(lambda x : x.strftime('%Y-%m'))
train['Month'] = train.date.dt.month
train['Day'] = train.date.dt.day
```

```python
# Lets check the profile again
ProfileReport(train, title="Dataframe Train Report", explorative=True)
```
