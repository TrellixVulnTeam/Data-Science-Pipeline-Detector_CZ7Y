**About**

The aim of this analysis is to perform the exploratory analysis of train dataset initially and get some insights on how each of the variables relate to our target variable which is 'sales'. Further we will use the other datasets along with train and get indepth insights.

We have about 06 atributes (including the target)

```python
sales_train = pd.read_csv('../input/store-sales-time-series-forecasting/train.csv')
print(sales_train.describe())
sales_train

```

```python
s=["date","family"]
print(sales_train[s].describe())
```

1. There is no missing data.
2. There are 33 unique "family" types, of which "AUTOMOTIVE" occurs the most (9k times)
3. Data is taken over 1864 days.


```python
plt.figure(figsize=(25,8))
sns.barplot(x="family", y="sales", data=sales_train)
plt.xticks(rotation=90)
```

Although "AUTOMOTIVE" occured the most the sales of automotive is not much compared to other "family".
"GROCERY I" , "PRODUCE" , "BEVERAGES" are the top 3 "family" that have most sales.

```python
plt.figure(figsize=(25,8))
sns.barplot(x="family", y="onpromotion", data=sales_train)
plt.xticks(rotation=90)
```

"GROCERY I" , "PRODUCE" , "BEVERAGES" are the top 3 "family" that have most number of items on promotion
From the previous and the present barplots we can see that there is a relation between the "sales" and "onpromotion"


Let us confirm the co-relations.

```python
tempd=sales_train[['date', 'store_nbr', 'family', 'sales', 'onpromotion']]


plt.figure(figsize=(6,6))
mask = np.triu(np.ones_like(tempd.corr(), dtype=bool))
sns.heatmap(tempd.corr(), mask = mask, cmap="YlGnBu", annot=True)
```

Hence we can conclude that the items/"family" that are on promotion have most sales.
