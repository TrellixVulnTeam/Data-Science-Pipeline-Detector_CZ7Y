### [Store Sales - Time Series Forecasting](https://www.kaggle.com/c/store-sales-time-series-forecasting): Naive one-day model
In this model we simply assume that tomorrows sales will be the the same as the sales today.

> "*The simplest representations of prior knowledge in forecasting
models are no-change models. Forecasts from appropriately formulated
no-change models are hard to beat in many forecasting situations*" <br>
[Kesten C. Green and J. Scott Armstrong "*Simple versus complex forecasting: The evidence*", Journal of Business Research **Volume 68** pp. 1678-1685 (2015)](https://doi.org/10.1016/j.jbusres.2015.03.026)

```python
import pandas as pd

# read in the data
train = pd.read_csv("../input/store-sales-time-series-forecasting/train.csv")
test  = pd.read_csv("../input/store-sales-time-series-forecasting/test.csv")

# select the very last day of the training data
last_day = train.query("date == '2017-08-15' ").reset_index(drop = True)

# store the values
last_day_value = last_day.groupby(['store_nbr', 'family'])['sales'].mean().to_dict()
test["sales"]  = test.set_index(['store_nbr', 'family']).index.map(last_day_value.get)

# create and write out the submission.csv file
submission = pd.DataFrame({'id': test.id, 'sales': test.sales})
submission.to_csv('submission.csv', index=False)
```
