****Train data:**** The games.csv contains match information. The key variable is gameId. 

Id: identifier, unique (numeric)

Date: date in year-month-day format 

family: 33 unique values (category)

Sales:  Sales value( float64)

onpromotion:  float64


```python
train = pd.read_csv('/kaggle/input/store-sales-time-series-forecasting/train.csv')
train.head()
```

```python
train.info()
```

```python
#check = train['date'].value_counts().reset_index()
check=train.groupby('date').agg({"sales" : "mean"}).reset_index()

check.columns = [
    'date', 
    'sales'
]
check = check.sort_values('sales')

fig = px.bar(
    check, 
    y='date', 
    x="sales", 
    color="sales",
    orientation='h', 
    title='Average Sales for every date', 
    height=800, 
    width=800
)
fig.show()


#check = train['date'].value_counts().reset_index()
check=train.groupby('family').agg({"sales" : "mean"}).reset_index()

check.columns = [
    'family', 
    'sales'
]
check = check.sort_values('sales')

fig = px.bar(
    check, 
    y='family', 
    x="sales", 
    color="sales",
    orientation='h', 
    title='Average Sales for every family', 
    height=800, 
    width=800
)
fig.show()



#check = train['date'].value_counts().reset_index()
check=train.groupby('store_nbr').agg({"sales" : "mean"}).reset_index()

check.columns = [
    'store_nbr', 
    'sales'
]
check = check.sort_values('sales')

fig = px.bar(
    check, 
    y='store_nbr', 
    x="sales", 
    color="sales",
    orientation='h', 
    title='Average Sales for every store number', 
    height=800, 
    width=800
)
fig.show()


#check = train['date'].value_counts().reset_index()
check=train.groupby('onpromotion').agg({"sales" : "mean"}).reset_index()

check.columns = [
    'onpromotion', 
    'sales'
]
check = check.sort_values('sales')

fig = px.bar(
    check, 
    y='onpromotion', 
    x="sales", 
    color="sales",
    orientation='h', 
    title='Average Sales on promotion', 
    height=800, 
    width=800
)
fig.show()
```

```python

check = train['family'].value_counts().reset_index()

check.columns = [
    'family', 
    'train'
]

check = check.sort_values('family')

fig = px.pie(
    check, 
    names='family', 
    values="train",  
    title='Family Variable',
    height=500,
    width=800
)

fig.show()
```
