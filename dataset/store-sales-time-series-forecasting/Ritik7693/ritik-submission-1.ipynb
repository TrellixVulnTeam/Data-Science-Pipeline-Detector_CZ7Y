```python


print('There are {} days, {} stores and {} families in the training data'.format(trainset['date'].nunique(),
                                                                                trainset['family'].nunique(),
                                                                                trainset['store_nbr'].nunique()))

```

sales by family

```python
family_daily=trainset.groupby(['date','family']).agg({'sales':'sum'}).reset_index()
family_daily.rename(columns={'sales':'family_daily_sales'},inplace=True)
print(family_daily)
```

lets see total daily time series

```python
total_daily=trainset.groupby(['date']).agg({'sales':'sum'}).reset_index()
total_daily.rename(columns={'sales':'total_daily_sales'},inplace=True)
print(total_daily.head())
```

**trining data by stores**

```python
store_daily=trainset.groupby(['date','store_nbr']).agg({'sales':'sum'}).reset_index()
store_daily.rename(columns={'sales':'store_daily_sales'},inplace=True)
print(store_daily)
```
