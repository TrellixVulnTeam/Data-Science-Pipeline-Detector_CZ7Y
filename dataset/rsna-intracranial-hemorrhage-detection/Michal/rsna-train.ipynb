```python
import torch
import torch.optim as optim
from model import EfficientFPN
from model import load_net
from train import Trainer


model = EfficientFPN(encoder_name='efficientnet-b3', classes=4, use_pretrained=False, 
                         grayscale=True) #TODO - evaluate losses

state = torch.load('../input/rsna-models/b3-uco-sasa_525.pth', map_location=lambda storage, loc: storage)
load_net(state, model)
#model.load_state_dict(state["state_dict"])
device = torch.device("cuda:0")
model = model.to(device)
optimizer = optim.SGD(model.parameters(), lr=1e-5, weight_decay=1e-5)


data_folder = "../input/understanding_cloud_organization"

model_trainer = Trainer(model, batch_size = {"train": 8, "val": 8}, 
                        data_folder=data_folder, optimizer = optimizer, 
                        train_width=512, train_height=320, run_dir='.', base_lr=1e-5,
                        num_epochs = 20, num_workers=8, grayscale=True,
                        from_epoch=0, finetune = False, use_tqdm=False, tensorboard=False)
model_trainer.start()
```
