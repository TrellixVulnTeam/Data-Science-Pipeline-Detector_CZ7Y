Let's see the label for each sub-type.

```python
train['Image ID'] = train['ID'].apply(lambda x: x.split('_')[1])
train['Sub-type'] = train['ID'].apply(lambda x: x.split('_')[2])
train = pd.pivot_table(train, index='Image ID', columns='Sub-type')
```

```python
train.head()
```

```python
train.mean()
```

```python
sample['Sub-type'] = sample['ID'].apply(lambda x:x.split('_')[2])
sample.loc[sample['Sub-type'] == 'any', 'Label'] = train.mean()[0]
sample.loc[sample['Sub-type'] == 'epidural', 'Label'] = train.mean()[1]
sample.loc[sample['Sub-type'] == 'intraparenchymal', 'Label'] = train.mean()[2]
sample.loc[sample['Sub-type'] == 'intraventricular', 'Label'] = train.mean()[3]
sample.loc[sample['Sub-type'] == 'subarachnoid', 'Label'] = train.mean()[4]
sample.loc[sample['Sub-type'] == 'subdural', 'Label'] = train.mean()[5]
```

```python
sample = sample.drop('Sub-type', axis=1)
```

```python
sample.to_csv('submit.csv', index=False)
```
