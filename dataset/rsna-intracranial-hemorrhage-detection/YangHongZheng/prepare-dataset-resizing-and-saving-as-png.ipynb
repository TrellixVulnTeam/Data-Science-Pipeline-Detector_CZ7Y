# Get images paths

```python
data_dir = "../input/rsna-intracranial-hemorrhage-detection/rsna-intracranial-hemorrhage-detection"
!ls {data_dir}
```

```python
train_dir = "stage_2_train"
train_paths = glob.glob(f"{data_dir}/{train_dir}/*.dcm")
print(len(train_paths))
```

```python
def get_patient_id(img_path):
    img_dicom = pydicom.read_file(img_path)
    return img_dicom.PatientID
```

```python
get_patient_id(train_paths[0])
```

```python
from collections import defaultdict 
from tqdm import tqdm
import random
dict_id = defaultdict(list)
random.shuffle(train_paths)
#print(train_paths)
for img_path in tqdm(train_paths):
    patient_id = get_patient_id(img_path)
    path = img_path[101:114] + 'png'
    dict_id[patient_id].append(path)
len(dict_id)
```
