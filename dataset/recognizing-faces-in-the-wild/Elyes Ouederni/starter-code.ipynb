```python
l =[]
for fam in tqdm(os.listdir("../input/recognizing-faces-in-the-wild/train/")):
    for person in os.listdir("../input/recognizing-faces-in-the-wild/train/"+fam):
        l.append(fam+'/'+person)
```

```python
len(l)
```

## Here I've used product function to generate all possible combinations betwen indeviduals

```python
from itertools import product
```

```python
matrix = []
while(len(l)!=1):
    p = l[0]
    l.remove(p)
    matrix.append(np.array(list(product([p],l)), dtype='object'))
matrix = pd.DataFrame(np.vstack(matrix), columns=['p1','p2'])
```

```python
matrix.head()
```

## Now we have all possible combination we should mark pairs mentionned in the train file as True else False

```python
train['target']=1
matrix = pd.merge(matrix, train, on=['p1','p2'], how='left')
```

```python
matrix['target'] = (matrix['target']
                            .fillna(0)
                            .astype(np.int8))
```
