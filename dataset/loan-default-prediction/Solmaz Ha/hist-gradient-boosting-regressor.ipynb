```python
df_train = pd.read_csv("../input/loan-default-prediction/train_v2.csv.zip")
df_test = pd.read_csv("../input/loan-default-prediction/test_v2.csv.zip")
```

# Cleaning The Data

```python
df_train.isnull().sum()
```

```python
df_train.dropna(inplace=True)
df_test.dropna(inplace=True)
```

```python
df_train.isnull().sum()
```

```python
df_train.shape
```

```python
for i in df_train.select_dtypes(include=['object']).columns:
    df_train.drop(labels=i, axis=1, inplace=True)
for i in df_test.select_dtypes(include=['object']).columns:
    df_test.drop(labels=i, axis=1, inplace=True)
```

```python
df_train.shape
```

```python
X = df_train.iloc[:,1:751]
y = df_train.iloc[:,751]
```

# Split data to training set and testing set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=14)
```

# Hist Gradient Boosting Regressor

```python
# explicitly require this experimental feature
from sklearn.experimental import enable_hist_gradient_boosting  # noqa
# now you can import normally from ensemble
from sklearn.ensemble import HistGradientBoostingRegressor
```

```python
est = HistGradientBoostingRegressor().fit(X_train, y_train)
```

```python
est.score(X_test, y_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(y_test, est.predict(X_test))
print("The mean squared error (MSE) on test set: {:.4f}".format(mse))
```
