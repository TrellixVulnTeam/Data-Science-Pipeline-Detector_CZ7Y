```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import ShuffleSplit

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
```

```python
models = {}
models['Logistic Regression'] = LogisticRegression()
models['KNN'] = KNeighborsClassifier()
models['Decision Tree'] = DecisionTreeClassifier()
models['Random Forest'] = RandomForestClassifier()
models['Naive Bayes'] = GaussianNB()
models['SVM'] = SVC()
```

```python
mean_acc = {}
results = []
for model in models:
    cv = ShuffleSplit(n_splits=5, test_size=.2)
    cv_results = cross_val_score(models[model],X_train, np.ravel(y_train),cv=cv,scoring='accuracy')
    results.append(cv_results)
    mean_acc[model] = round(cv_results.mean(), 3)

mean_acc
```

```python
from sklearn.preprocessing import StandardScaler, Normalizer

norm = Normalizer()
normalized_data = norm.fit_transform(X_train)
```

```python
mean_acc = {}
results = []
for model in models:
    cv = ShuffleSplit(n_splits=5, test_size=.2)
    cv_results = cross_val_score(models[model],normalized_data, np.ravel(y_train),cv=cv,scoring='accuracy')
    results.append(cv_results)
    mean_acc[model] = round(cv_results.mean(), 3)

mean_acc
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=10)
pca_data = pca.fit_transform(X_train)
pca_normalized_data = pca.fit_transform(normalized_data)
```

```python
mean_acc = {}
results = []
for model in models:
    cv = ShuffleSplit(n_splits=5, test_size=.2)
    cv_results = cross_val_score(models[model],pca_data, np.ravel(y_train),cv=cv,scoring='accuracy')
    results.append(cv_results)
    mean_acc[model] = round(cv_results.mean(), 3)

mean_acc
```

```python
mean_acc = {}
results = []
for model in models:
    cv = ShuffleSplit(n_splits=5, test_size=.2)
    cv_results = cross_val_score(models[model],pca_normalized_data, np.ravel(y_train),cv=cv,scoring='accuracy')
    results.append(cv_results)
    mean_acc[model] = round(cv_results.mean(), 3)

mean_acc
```

```python
model = KNeighborsClassifier()
model.fit(normalized_data, y_train.values.ravel())
```

```python
normalized_test = norm.fit_transform(X_test)
pca_normalized_test = pca.fit_transform(normalized_test)
prediction = model.predict(normalized_test)
```
