```python
train = pd.read_csv('/kaggle/input/data-science-london-scikit-learn/train.csv',header=None)
```

```python
test = pd.read_csv('/kaggle/input/data-science-london-scikit-learn/trainLabels.csv',header = None)
```

```python
train.info()
```

```python
train.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
corrmat = train.corr()
plt.subplots(figsize=(12,9))
sns.heatmap(corrmat,vmax=0.9)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

X_train,X_test,y_train,y_test = train_test_split(train,test,test_size=0.3)
```

```python
RandF = RandomForestClassifier(n_estimators=100)
RandF.fit(X_train,y_train.values.ravel())
y_pred = RandF.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
clf_lg = LogisticRegression()
clf_lg.fit(X_train,y_train.values.ravel())
lg_pred = clf_lg.predict(X_test)
```

```python
print(accuracy_score(y_test,lg_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
testL = pd.read_csv('/kaggle/input/data-science-london-scikit-learn/test.csv',header=None)
clf_knn = KNeighborsClassifier()
clf_knn.fit(X_train,y_train.values.ravel())
K_pred = clf_knn.predict(X_test)
```

```python
print(accuracy_score(K_pred,y_test))
```

```python
submit = clf_knn.predict(testL)
submission = pd.DataFrame(submit)
print(submission)
submission.columns = ['Solution']
print(submission)
submission['Id'] = np.arange(1,submission.shape[0]+1)
print(submission)
submission = submission[['Id','Solution']]
```
