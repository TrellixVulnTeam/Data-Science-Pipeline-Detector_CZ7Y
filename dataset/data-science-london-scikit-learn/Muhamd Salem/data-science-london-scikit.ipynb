```python
#import libry
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score, train_test_split
import warnings
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report
warnings.filterwarnings('ignore')
%matplotlib inline

```

```python
#reading_data
train = pd.read_csv('../input/data-science-london-scikit-learn/train.csv', header=None)
trainLabel = pd.read_csv('../input/data-science-london-scikit-learn/trainLabels.csv', header=None)
test = pd.read_csv('../input/data-science-london-scikit-learn/test.csv', header=None)
```

```python
X, y = train, np.ravel(trainLabel)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
#modeal_complex 
n_egl=np.arange(4,30)
kfold=10
train_acyorcy=[]
val_acyorcy=[]
best_score=0
best_score2=0
best_knn=None
best_knn2=None
for i,k in  enumerate(n_egl):
  knn=KNeighborsClassifier(n_neighbors=k)
  knn.fit(X_train,y_train)
  train_acyorcy.append(knn.score(X_train,y_train))
  val_acyorcy.append(np.mean(cross_val_score(knn,X,y,cv=kfold)))
  if np.mean(cross_val_score(knn,X,y,cv=kfold)) > best_score:
             best_score=np.mean(cross_val_score(knn,X,y,cv=kfold))
             best_knn = knn
  elif knn.score(X_train,y_train)>best_score2:
     best_score2=knn.score(X_train,y_train)   
     best_knn2=knn


```

```python
# Plot
plt.figure(figsize=[13,8])
plt.plot(n_egl, val_acyorcy, label = 'Validation Accuracy')
plt.plot(n_egl, train_acyorcy, label = 'Training Accuracy')
plt.legend()
plt.title('k value VS Accuracy')
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy')
plt.xticks(n_egl)
plt.show()

print('Best Accuracy without feature scaling:',best_score)
print('Best Accuracy without feature scaling2:',best_score2)
print(best_knn2)
print(best_knn)
```

```python
from sklearn.preprocessing import StandardScaler,minmax_scale,Normalizer
std=StandardScaler()
x_std=std.fit_transform(X)
min_x=minmax_scale(X)
Nor=Normalizer()
Nor_x=Nor.fit_transform(X)
```

```python
#complex_model
n_gl=np.arange(4,40)
k_fold=20
val_acurcy={'x_std':[],'min':[],'Nor_x':[]}
best_KNN=None
best_scaling=None
best_accurcy_ascaling=0
for i,k in enumerate(n_gl):
  KNN=KNeighborsClassifier(n_neighbors=k)
  S1=np.mean(cross_val_score(KNN,x_std,y,cv=k_fold))
  val_acurcy['x_std'].append(S1)
  S2=np.mean(cross_val_score(KNN,min_x,y,cv=kfold))
  val_acurcy['min'].append(S2)
  S3=np.mean(cross_val_score(KNN,Nor_x,y,cv=kfold))
  val_acurcy['Nor_x'].append(S3)
  if S1>best_accurcy_ascaling :
    best_KNN=KNN
    best_accurcy_ascaling=S1
    best_scaling='x_std'
  elif S2>best_accurcy_ascaling:
    best_KNN=KNN
    best_accurcy_ascaling=S2
    best_scaling='min_x'
  elif S3> best_accurcy_ascaling:
    best_KNN=KNN
    best_accurcy_ascaling=S3
    best_scaling='Nor_x'  
```

```python
#plot
plt.figure(figsize=[12,7])
plt.plot(n_gl,val_acurcy['x_std'],label=' Acurcy with StandardScaler')
plt.plot(n_gl,val_acurcy['min'],label='ACURCY withe minmax_scale ')
plt.plot(n_gl,val_acurcy['Nor_x'],label='Acurcy withe Normalizer')
plt.legend()
plt.xlabel('knn_numper')
plt.ylabel('Acurcy')
plt.title('KNN & Accurcy')
plt.xticks(n_gl)
plt.show()
print('Best Accuracy with feature scaling:', best_KNN)
print('Best kNN classifier:', best_scaling)
print('Best scaling:', best_accurcy_ascaling)
```

```python
#Calculating Confusion Matrix
CM = confusion_matrix(y_test, best_KNN.predict(X_test))
print('Confusion Matrix is : \n', CM)

# drawing confusion matrix
sns.heatmap(CM, center = True)
plt.show()
```

```python
#Calculating classification Report :  
classification_report(y_test,  best_KNN.predict(X_test), labels=None, target_names=None,sample_weight=None, digits=2, output_dict=False)

ClassificationReport = classification_report(y_test, best_KNN.predict(X_test))
print('Classification Report is : ', ClassificationReport )
```
