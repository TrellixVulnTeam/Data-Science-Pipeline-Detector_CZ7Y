**Train-Test Split**

```python
from sklearn.model_selection import train_test_split

# Train-Test split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.30, random_state = 123)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

**Train various models**

```python
from sklearn.metrics import accuracy_score 
from sklearn.model_selection import cross_val_score

# NAIBE BAYES
from sklearn.naive_bayes import GaussianNB

model_1 = GaussianNB()
model_1.fit(X_train,y_train.values.ravel()) #values.ravel
predicted_1 = model_1.predict(X_test)
print('Naive Bayes',accuracy_score(y_test, predicted_1)) #accuracy score

#KNN
from sklearn.neighbors import KNeighborsClassifier

model_2 = KNeighborsClassifier()
model_2.fit(X_train,y_train.values.ravel()) 
predicted_2 = model_2.predict(X_test)
print('KNN',accuracy_score(y_test, predicted_2))

#RANDOM FOREST
from sklearn.ensemble import RandomForestClassifier

model_3 = RandomForestClassifier(n_estimators = 100,random_state = 99) #n_estimators, random_state
model_3.fit(X_train,y_train.values.ravel())
predicted_3 = model_3.predict(X_test)
print('Random Forest',accuracy_score(y_test,predicted_3))

#LOGISTIC REGRESSION
from sklearn.linear_model import LogisticRegression

model_4 = LogisticRegression(solver = 'saga') #solver
model_4.fit(X_train,y_train.values.ravel())
predicted_4 = model_4.predict(X_test)
print('Logistic Regression',accuracy_score(y_test, predicted_4))

#SVM
from sklearn.svm import SVC

model_5 = SVC(gamma = 'auto') #gamma
model_5.fit(X_train,y_train.values.ravel())
predicted_5 = model_5.predict(X_test)
print('SVM',accuracy_score(y_test, predicted_5))

#DECISON TREE
from sklearn.tree import DecisionTreeClassifier

model_6 = DecisionTreeClassifier()
model_6.fit(X_train,y_train.values.ravel())
predicted_6 = model_6.predict(X_test)
print('Decision Tree',accuracy_score(y_test, predicted_6))

#XGBOOST
from xgboost import XGBClassifier

model_7 = XGBClassifier(use_label_encoder=False) #use_label_encoder=False
model_7.fit(X_train,y_train.values.ravel())
predicted_7 = model_7.predict(X_test)
print('XGBoost',accuracy_score(y_test, predicted_7))
```
