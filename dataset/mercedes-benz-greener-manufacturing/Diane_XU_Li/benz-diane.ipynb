```python
import pandas as pd
from pycaret.regression import *
```

```python
train = pd.read_csv("../input/mercedes-benz-greener-manufacturing/train.csv.zip")
test = pd.read_csv("../input/mercedes-benz-greener-manufacturing/test.csv.zip")
```

```python
train = train.sample(frac=0.5)
train.head()
```

```python
reg = setup(data = train, target = 'y', numeric_imputation = 'mean',
            categorical_features = ['X0','X1','X2','X3','X4','X5','X6','X8'], ignore_features = ['ID'],
            silent = True)
```

```python
compare_models()
```

```python
gbr = create_model('gbr')
```

```python
tune_gbr = tune_model(gbr)
```

```python
plot_model(tune_gbr)
```

```python
plot_model(tune_gbr, plot="learning")
```

```python
plot_model(tune_gbr, plot="feature")
```

```python
plot_model(tune_gbr, plot="learning")
```

```python
br = create_model('br')
tune_br = tune_model(br)
```

```python
rf = create_model('rf')
tune_rf = tune_model(rf)
```

```python
lgbm = create_model('lightgbm')
tune_lgbm = tune_model(lgbm)
```

```python
blended = blend_models(estimator_list = [tune_gbr, tune_br, tune_rf, tune_lgbm], 
                       fold = 5)
```

```python
predictions = predict_model(blended, data = test)
predictions.head()
```
