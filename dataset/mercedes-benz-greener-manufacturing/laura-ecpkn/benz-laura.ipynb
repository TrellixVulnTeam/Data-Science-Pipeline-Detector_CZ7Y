```python
import pandas as pd
from pycaret.regression import *
```

```python
train = pd.read_csv("../input/mercedes-benz-greener-manufacturing/train.csv.zip")
test = pd.read_csv("../input/mercedes-benz-greener-manufacturing/test.csv.zip")
```

```python
train = train.sample(frac=0.3)
train
```

```python
reg = setup(data = train, 
            target = 'y',
            numeric_imputation = 'mean',
            categorical_features = ['X0','X1','X2','X3','X4','X5','X6','X8'], 
            ignore_features = ['ID'],
            silent = True)
```

```python
compare_models()
```

```python
br = create_model('br')
```

```python
tune_br = tune_model(br)
```

```python
plot_model(tune_br)
```

```python
plot_model(tune_br, plot="error")
```

```python
plot_model(tune_br, plot="learning")
```

```python
plot_model(tune_br, plot="feature")
```

```python
huber = create_model('huber')
tune_huber = tune_model(huber)
```

```python
gbr = create_model('gbr')
tune_gbr = tune_model(gbr)
```

```python
lgbm = create_model('lightgbm')
tune_lgbm = tune_model(lgbm)
```

```python
blended = blend_models(estimator_list = [tune_lgbm, tune_huber, tune_br, tune_gbr], 
                       fold = 5)
```

```python
predictions = predict_model(blended, data = test)
predictions.head()
```
