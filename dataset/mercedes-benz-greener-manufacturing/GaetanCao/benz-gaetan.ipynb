```python
import pandas as pd
test = pd.read_csv("../input/mercedes-benz-greener-manufacturing/test.csv.zip")
train = pd.read_csv("../input/mercedes-benz-greener-manufacturing/train.csv.zip")
```

```python
train
```

```python
from pycaret.regression import *
```

```python
reg = setup(data=train, target="y" )
```

```python
models = compare_models()
```

```python
br=create_model('br')
```

```python
omp=create_model('omp')
```

```python
lgbm=create_model("lightgbm")
```

```python
tune_lgbm=tune_model(lgbm)
```

```python
plot_model(tune_lgbm)
```

```python
plot_model(tune_lgbm,plot='feature')
```

```python
plot_model(tune_lgbm,plot='error')
```

```python
plot_model(tune_lgbm,plot='learning')
```

```python
blended=blend_models(estimator_list=[tune_lgbm,omp,br],fold=5,choose_better=True)
```

```python
predictions=predict_model(blended,data=test)
```

```python
submission=pd.read_csv('../input/mercedes-benz-greener-manufacturing/sample_submission.csv.zip')
```

```python
submission['y']=predictions['Label']
```
