```python
import pandas as pd
from pycaret.regression import *
```

```python
train = pd.read_csv("../input/mercedes-benz-greener-manufacturing/train.csv.zip")
test = pd.read_csv("../input/mercedes-benz-greener-manufacturing/test.csv.zip")
```

```python
reg = setup(data=train, target='y',imputation_type='iterative')
```

```python
models = compare_models()
```

```python
omp=create_model('omp')
```

```python
tune_omp=tune_model(omp)
```

```python
plot_model(tune_omp)
```

```python
plot_model(tune_omp,plot='error')
```

```python
plot_model(tune_omp,plot='learning')
```

```python
plot_model(tune_omp,plot='feature')
```

```python
gbr=create_model('gbr')
```

```python
tune_gbr=tune_model(gbr)
```

```python
br=create_model('br')
```

```python
tune_br=tune_model(br)
```

```python
blended=blend_models(estimator_list=[tune_gbr,tune_omp,tune_br],fold=5,choose_better=True)
```

```python
stacked=stack_models(estimator_list=[tune_omp,tune_br],meta_model=tune_gbr,restack=False)
```

```python
plot_model(stacked,plot='error')
```

```python
predictions=predict_model(stacked,data=test)
predictions.head()
```

```python
sample_submission=pd.read_csv('/kaggle/input/mercedes-benz-greener-manufacturing/sample_submission.csv.zip',
                              compression='zip')
```

```python
sample_submission['y']=predictions['Label']
sample_submission.to_csv('submission.csv',index=False)
```
