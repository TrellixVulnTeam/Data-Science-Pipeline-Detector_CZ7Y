We can also use mode instead of using df[var].value_counts().index[0]

```python
## Replacing Function
def impute_nan(df,var):
    freq_category = df[var].value_counts().index[0]
    df[var].fillna(freq_category, inplace = True)
    
```

```python
for feature in ['BsmtQual','FireplaceQu','GarageType']:
    impute_nan(df,feature)
```

```python
df.isnull().mean()
```

# Target Guided Ordinal Encoding 
Ordering the Lables according to Target
Replace The labels by the joint probability of being 1 or 0


```python
df = pd.read_csv('../input/titanic/train.csv', usecols = ['Cabin','Survived'])
```

```python
df.head()
```

```python
df['Cabin'].fillna("Missing", inplace = True) #Replacing NAN WITH LABEL
```

```python
df.head()
```

```python
df['Cabin'] = df['Cabin'].astype(str).str[0]
df.Cabin.unique()
#WE HAVE CREATED TARGET VARIABLES.
```

```python
ordinal_labels = df.groupby(['Cabin'])['Survived'].mean().sort_values().index #each coulumns how many group survived.
```

```python
#dictionary comprehension.
i = 0
dictionary = {} 
ordinal_labels = list(ordinal_labels)
while i < len(ordinal_labels):
    dictionary[ordinal_labels[i]] = i 
    i = i+1
```

```python
dictionary
```

```python
df['Cabin_Ordinal'] = df['Cabin'].map(dictionary)
```

# Probablity Ratio Encoding.
1. Probability of Survived base on cabin --Categorical Feature
2. Probablity of Dead base on Survived as total should be 1 so subtracting 1 from survived gave us dead probablity 
3. We Will calculate ratio of survived and dead and this ratio will give us values for our categories to use probablity encoding.
4. Dictionary to map cabin with probablity_ratio 
4. replace categorical feature with encoded values.

```python
df = pd.read_csv('../input/titanic/train.csv', usecols = ['Cabin','Survived'])
```

```python
df['Cabin'].fillna("Missing", inplace = True) #Replacing NAN WITH LABEL
```

```python
df.Cabin.unique() #LETS CONVERT THEM 
```

```python
df['Cabin'] = df['Cabin'].astype(str).str[0]
df.head()
```

```python
df.Cabin.unique() # WALLAH NOW WE DON't Have To Deal With Bigger Values
```

```python
df.groupby(['Cabin'])['Survived'].mean() # LETS GET Percentage of survived
```

```python
#THIS PERCENTAGE CAN BE USED AS OUR PROBABLITY
prob_df = df.groupby(['Cabin'])['Survived'].mean()
prob_df = pd.DataFrame(prob_df)
```

```python
prob_df # THIS IS OUR PROBABLITY OF SURVIVAL ON THE BASIS OF CABIN
```

```python
prob_df['Died'] = 1-prob_df['Survived'] #AND THIS  IS DEAD PROBABLITY.
```

```python
prob_df
```

# SO WHAT IS PROBABLITY ENCODING : WE WILL SAY PROBABLITY OF ONE FEATURE / PROBABLITY OF OTHER FEATURE MADE THROUGH THAT FEATURE AND ENCDOING THAT NO 

```python
prob_df['Probablity_ratio'] = prob_df['Survived']/prob_df['Died']
prob_df.head()
```

Now this ratio can be converted To my category of cabin and we are done
with encoding

```python
prob_encode = prob_df['Probablity_ratio'].to_dict()
```

```python
df['Cabin_encoded'] = df['Cabin'].map(prob_encode) # MAPPING ENCODE VAL
```

```python
df.head()
```

## If you want to check wether feature is guassian or normal distribution
We use Q-Q plot technique

```python
#qqplot technique (THIS IS PRETTY IMPORTANT)

import scipy.stats as stat
def plot_data(df,feature):
    plt.figure(figsize = (10,6))
    plt.subplot(1,2,1)
    df[feature].hist()
    plt.subplot(1,2,2)
    stat.probplot(df[feature], dist = 'norm', plot = pylab)
    plt.show()
```

```python
plot_data(df,'Age')
```

# if its falling in the same line age is gaussian distributed 
as you can see above 

```python
#  Lets try logarthmic Transformation and see if we get same 
#     gaussian distribution 
df['Age_log'] = np.log(df['Age'])
plot_data(df,'Age_log')
```

#### SO NO Its NOT SUITABLE TO USE IT FOR AGE to do logarthmic Transformation

**Reciprocal** **Transformation**

```python
#lets try reciprocal transformation
df['Age_reci']  = 1/df['Age']
plot_data(df,'Age_reci')
#AGAIN NO WE SHOULD NOT USE THIS DATA IS NOT Gaussian
#Distributed
```

```python
#lets Try Square Root Transformation
df['Age_root'] = df.Age**(1/2) # TO THE POWER OF 1/2
plot_data(df,'Age_root')
#YES THIS IS GAUSSIAN DISTRIBUTED AND WE CAN USE THIS.
```

```python
#lets try the same thing with Exponential Transformation
df['Age_exponential'] = df.Age**(1/1.2) # POWER OF 1/1.2
plot_data(df, 'Age_exponential')
```

### BoxCox Tansformation
T(Y) = (Y exp(LAMBDA) -1)/LAMBDA

where Y is the response variable and lambda is the transformation parameter. lambda varies from -5 to 5. In the Transformation, all values of lambda are considered and the optimal value for a given variable is selected.

```python
df['Age_Boxcox'],parameters = stat.boxcox(df['Age'])
```

```python
print(parameters)
```

```python
plot_data(df,'Age_Boxcox')
```
