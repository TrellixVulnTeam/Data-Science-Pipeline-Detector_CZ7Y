```python
import warnings
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import Normalizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
warnings.filterwarnings('ignore')
```

```python
data_train=pd.read_csv('../input/santander-customer-transaction-prediction/train.csv')
data_test=pd.read_csv('../input/santander-customer-transaction-prediction/test.csv')
```

```python
sns.distplot(data_train['target'])
```

```python
y_train=data_train['target']
x_train=data_train.drop(['ID_code','target'],axis=1)
data_test=data_test.drop(['ID_code'],axis=1).copy()

```

```python
#Normalizing Data

scaler = Normalizer(copy=True, norm='max') # you can change the norm to 'l1' or 'max' 
x_train = scaler.fit_transform(x_train)

#showing data
print('X \n' , x_train[:10])
```

```python
#Splitting data

X_train, X_test, y_train, y_test = train_test_split(x_train, y_train, test_size=0.33, random_state=44, shuffle =True)

#Splitted Data
print('X_train shape is ' , X_train.shape)
print('X_test shape is ' , X_test.shape)
print('y_train shape is ' , y_train.shape)
print('y_test shape is ' , y_test.shape)
```

```python
#Applying LogisticRegression Model 

'''
linear_model.LogisticRegression(penalty='l2’,dual=False,tol=0.0001,C=1.0,fit_intercept=True,intercept_scaling=1,
                                class_weight=None,random_state=None,solver='warn’,max_iter=100,
                                multi_class='warn’, verbose=0,warm_start=False, n_jobs=None)
'''

LogisticRegressionModel = LogisticRegression(penalty='l2',solver='liblinear',C=1.0,random_state=33)
LogisticRegressionModel.fit(X_train, y_train)

#Calculating Details
print('LogisticRegressionModel Train Score is : ' , LogisticRegressionModel.score(X_train, y_train))
print('LogisticRegressionModel Test Score is : ' , LogisticRegressionModel.score(X_test, y_test))
print('LogisticRegressionModel Classes are : ' , LogisticRegressionModel.classes_)
print('LogisticRegressionModel No. of iteratios is : ' , LogisticRegressionModel.n_iter_)

#Calculating Prediction
y_pred = LogisticRegressionModel.predict(X_test)
data_pred=LogisticRegressionModel.predict(data_test)
print('Predicted Value for LogisticRegressionModel is : ' , y_pred[:10])
print('Predicted Value data_test for LogisticRegressionModel is : ' , y_pred[:10])
```

```python
#Calculating Confusion Matrix
CM = confusion_matrix(y_test, y_pred)
print('Confusion Matrix is : \n', CM)

# drawing confusion matrix
sns.heatmap(CM, center = True)
plt.show()
#Calculating classification Report :  
classification_report(y_test, y_pred, labels=None, target_names=None,sample_weight=None, digits=2, output_dict=False)

ClassificationReport = classification_report(y_test,y_pred)
print('Classification Report is : ', ClassificationReport )
```
