```python
from supervised.automl import AutoML 
```

```python
train = pd.read_csv("../input/santander-customer-transaction-prediction/train.csv")
test = pd.read_csv("../input/santander-customer-transaction-prediction/test.csv")
```

```python
train["target"].value_counts()
```

```python
y_train = train["target"]
X_train = train.drop(["target"], axis=1)
```

# Exploration des donn√©es (EDA)


```python
from supervised.preprocessing.eda import EDA
```

```python
EDA.extensive_eda(X_train,y_train,save_path=".")
```

```python
automl = AutoML(mode="Explain", eval_metric="accuracy")
```

```python
automl.fit(X_train, y_train)
```

```python
automl.report()
```

```python
predictions = automl.predict(test)
```

# soumission

```python
sample_submission = pd.read_csv("../input/santander-customer-transaction-prediction/sample_submission.csv")
```

```python
sample_submission['target'] = predictions
sample_submission.to_csv('submission.csv',index=False)
```
