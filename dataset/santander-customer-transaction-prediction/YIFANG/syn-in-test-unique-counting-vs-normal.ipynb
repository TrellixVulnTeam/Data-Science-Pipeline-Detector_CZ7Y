```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
print(os.listdir("../input"))
import  warnings
warnings.simplefilter('ignore')
```

```python
#######################################load data################################################################
train=pd.read_csv('../input/santander-customer-transaction-prediction/train.csv')
test=pd.read_csv('../input/santander-customer-transaction-prediction/test.csv')
pr_b=np.load('../input/list-of-fake-samples-and-public-private-lb-split/synthetic_samples_indexes.npy')
################################Split syn and true##############################################################
#train.drop(['ID_code','target'],axis=1,inplace=True)
test=test.drop('ID_code',axis=1)
# syn
#test1=test.iloc[pr_b]
# true
test2=test.drop(pr_b)
```

# True in test vs normal

```python
np.random.seed(9)
# true values unique number  VS random generated samples
for col in test2.columns:
    print(col)
    a=test2[col].map(test2[col].value_counts())
    sns.distplot(np.round(a,4),bins=100)
    plt.show()
    a=pd.DataFrame(np.round(np.random.normal(loc=np.mean(train[col]), scale=np.std(train[col]), size=100000),4))
    a.columns=['x']
    b=a.x.map(a.x.value_counts())
    sns.distplot(b,bins=100)
    plt.show()
    
```
