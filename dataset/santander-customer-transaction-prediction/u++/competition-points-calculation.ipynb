```python
import numpy as np


def calculate_point(n_teammates, rank, n_teams, t):
    point = (100000 / (n_teammates) ** 0.5) * (rank ** -0.75) * np.log10(1 + np.log10(n_teams)) * np.exp(-1 * t / 500) #Point calculation as of 17-04-2019
    return point
```

```python
n_teammates = 1 # number of teammates
rank = 1 # position landed on LB
n_teams = 1115 # number of teams 
t = 0 # t is the number of days elapsed since the point was awarded.
```

```python
# Solo competitor at #1
calculate_point(n_teammates=1,rank=1,n_teams=n_teams,t=t)
```

```python
# Solo competitor at #7
calculate_point(n_teammates=1,rank=7,n_teams=n_teams,t=t)
```
