```python
y = train.loc[:, 'target']
y.head()
```

```python
x = train.loc[:, :]
x.drop('target', inplace = True, axis = 1)
x.head()
```

```python
col = x.columns
std = StandardScaler()
x = std.fit_transform(x)
x
```

```python
x = pd.DataFrame(data = x, columns = col)
```

```python
balance = RandomOverSampler()
x, y = balance.fit_resample(x, y)
```

```python
print(y.value_counts())
plt.figure(figsize = (8,5))
sns.countplot(x = y)
plt.xlabel("Target", size = 12)
plt.ylabel("Count", size = 12)
plt.title("Distribution in target column after resampling", size = 12)
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.25, random_state = 42)
```

```python
light = LGBMClassifier(n_jobs=-1)
light.fit(xtrain, ytrain, eval_metric = 'auc', eval_set = (xtest, ytest), verbose = 10, early_stopping_rounds = 100)
```

```python
predicted = light.predict_proba(xtrain)[:,1].round()
train_score = accuracy_score(predicted, ytrain)*100
print("Accuracy using LightGBM on training data is {} %".format(train_score))
print("Roc auc score using LightGBM on training data is {} %".format(roc_auc_score(predicted, ytrain)))
```

```python
predicted = light.predict_proba(xtest)[:,1].round()
test_score = accuracy_score(predicted, ytest)*100
print("Accuracy using LightGBM testing data is {} %".format(test_score))
print("Roc auc score using LightGBM on testing data is {} %".format(roc_auc_score(predicted, ytest)))
```
