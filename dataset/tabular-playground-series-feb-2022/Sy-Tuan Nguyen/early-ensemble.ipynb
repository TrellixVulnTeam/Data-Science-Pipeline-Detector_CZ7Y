# **<span style="color:#61210B"> Early Ensemble</span>**

This notebook presents an automated ensembling using predicted results from the most relevant public notebooks. 

```python
import os
import numpy as np
import pandas as pd
from scipy import stats
import plotly.express as px
from sklearn.preprocessing import LabelEncoder
```

```python
targetName = 'target'
competitionDir = '/kaggle/input/tabular-playground-series-feb-2022'
submission = pd.read_csv('../input/tabular-playground-series-feb-2022/sample_submission.csv')
```

```python
train = pd.read_pickle("../input/tpsfeb2022-ds-to-pickle-with-folds/train.pkl")
le = LabelEncoder()
le.fit_transform(train.target)
```

```python
preds = []
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        if (dirname != competitionDir) & ((filename=='submission.csv') or (filename=='submission__blend.csv') or (filename=='lightautoml_tabularautoml.csv')):
            df = pd.read_csv(os.path.join(dirname, filename))
            #for i in range(np.random.randint(2,3)):
            preds.append(le.transform(df[targetName])) 
```

```python
submission[targetName] = le.inverse_transform(stats.mode(np.array(preds), axis=0)[0].transpose())
submission.to_csv("submission.csv", index=False)
```
