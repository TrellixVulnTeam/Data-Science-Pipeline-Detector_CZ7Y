# A simple code to generate feature names

Below is a simple code to generate exactly the feature names

```python
# Load data
import pandas as pd
train = pd.read_csv("../input/tabular-playground-series-feb-2022/train.csv")

# Feature names
cols = train.columns.tolist()

# A simple code to generate exactly the feature names
features=[]
for k in range(11):
    for j in range(11-k):
        for i in range(11-j-k):
            features.append(f'A{k}T{j}G{i}C{10-i-j-k}')
            
print('Number of generated feature names:', len(features))
print('The generated feature names:', features)

# Verify that the generated feature names match with the true feature names
for i in range(len(features)):
    assert(cols[i+1]==features[i])
```
