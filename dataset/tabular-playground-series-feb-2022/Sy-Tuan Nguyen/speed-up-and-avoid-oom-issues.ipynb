```python
import pandas as pd
import numpy as np
from sklearn.model_selection import StratifiedKFold

train = pd.read_csv('../input/tabular-playground-series-feb-2022/train.csv', index_col=0).drop_duplicates(keep='first').reset_index(drop=True)

test  = pd.read_csv('../input/tabular-playground-series-feb-2022/test.csv', index_col=0)
sub   = pd.read_csv('../input/tabular-playground-series-feb-2022/sample_submission.csv') 

X_train = train.drop('target', axis=1).astype('float32')
y_train = train.target
test  = test.astype('float32')

train = pd.concat([X_train, y_train], axis=1)

for n_split in [5, 10, 20]:
    skf = StratifiedKFold(n_split, shuffle=True, random_state=42)
    folds = np.zeros(len(y_train)).astype('int16')
    for fold, (trn_idx, val_idx) in enumerate(skf.split(X_train, y_train)):
        folds[val_idx]=fold
    
    train[f'{n_split}_folds'] = folds
    train[f'{n_split}_folds'] = train[f'{n_split}_folds'].astype('int16')
```
