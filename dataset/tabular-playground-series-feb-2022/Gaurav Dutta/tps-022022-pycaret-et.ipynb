```python
import pandas as pd
train=pd.read_csv('../input/tabular-playground-series-feb-2022/train.csv')
train.head()
```

```python
train=train.drop(['row_id'],axis=1)
train.head()
```

```python
train.describe()
```

```python
train.info()
```

```python
train.isnull().sum()
```

```python
test_fake=train.sample(frac=0.2) #workaround to make sure we train 100% of the train.csv than training 1% to 99% data
test_fake.reset_index(drop=True, inplace=True)
test_fake.head()
```

```python
from pycaret.classification import *
```

```python
dna= setup(data=train,
          target = "target",  session_id=42,
           remove_multicollinearity = True, 
          test_data = test_fake,silent=True,
           data_split_shuffle=True,data_split_stratify=True
         )
```

```python
et=create_model('et')
et
```

```python
test=pd.read_csv('../input/tabular-playground-series-feb-2022/test.csv')
test.head()
```

```python
test=test.drop(['row_id'],axis=1)
test.head()
```

```python
predictions=predict_model(et,data=test)
predictions.head()
```

```python
sol=pd.read_csv('../input/tabular-playground-series-feb-2022/sample_submission.csv')
sol.head()
```

```python
sol.target=predictions.Label
sol.head()
```

```python
sol.to_csv('./submission.csv',index=False)
```
