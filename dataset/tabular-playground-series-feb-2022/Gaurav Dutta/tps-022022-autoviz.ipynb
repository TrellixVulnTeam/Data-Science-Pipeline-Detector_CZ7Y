```python
import pandas as pd
train=pd.read_csv("../input/tabular-playground-series-feb-2022/train.csv")
train.head()
```

```python
train=train.drop(['row_id'],axis=1)
train.head()
```

```python
## downcasting loop
for column in train:
    if train[column].dtype == 'float64':
        train[column]=pd.to_numeric(train[column], downcast='float')
    if train[column].dtype == 'int64':
        train[column]=pd.to_numeric(train[column], downcast='integer')
train.head()
```

```python
import numpy as np
from sklearn.preprocessing import LabelEncoder
X = train.loc[:,:'A10T0G0C0']
y = train['target']
label_encoder = LabelEncoder()
y_le = label_encoder.fit_transform(y)
y= np.unique(y)
y_le = np.unique(y_le)
y_encode = dict(zip(y, y_le))
print(y_encode)
```

```python
train['target']=train['target'].map(y_encode)
train.head()
```

```python
col_list=train.columns.tolist()
n = 10
final = [col_list[i * n:(i + 1) * n] for i in range((len(col_list) + n - 1) // n )] 
final[0]
```

```python
for i in range(len(final)-1):
    final[i].append('target')
final[int(len(final)-1)]
```
