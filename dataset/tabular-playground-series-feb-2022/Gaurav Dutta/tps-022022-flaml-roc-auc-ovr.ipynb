```python
import pandas as pd
train=pd.read_csv("../input/tabular-playground-series-feb-2022/train.csv")
train.head()
```

```python
train.shape
```

```python
train.nunique()
```

```python
train=train.drop(['row_id'],axis=1)
train.head()
```

```python
train.info()
```

```python
round(train.target.value_counts()*100/len(train),2)
```

```python
y_train= train.pop('target')
X_train = train
```

```python
from flaml import AutoML
automl = AutoML()
```

```python
automl.fit(X_train, y_train, task="classification",metric='roc_auc_ovr',time_budget=36000)
```

```python
print('Best ML leaner:', automl.best_estimator)
print('Best hyperparmeter config:', automl.best_config)
print('Best roc_auc_ovr on validation data: {0:.4g}'.format(1-automl.best_loss))
print('Training duration of best run: {0:.4g} s'.format(automl.best_config_train_time))
```

```python
test=pd.read_csv("../input/tabular-playground-series-feb-2022/test.csv")
test.head()
```

```python
test=test.drop(['row_id'],axis=1)
test.head()
```

```python
y_pred= automl.predict(test)
y_pred
```

```python
df = pd.DataFrame(y_pred,columns=['target'])
df.head()
```

```python
sol=pd.read_csv('../input/tabular-playground-series-feb-2022/sample_submission.csv')
sol.head()
```

```python
sol.target=df.target
sol.head()
```

```python
sol.to_csv('./submission.csv',index=False)
```
