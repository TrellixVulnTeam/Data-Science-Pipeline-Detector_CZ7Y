```python
def give_me_a_row(df, idx_row):
    """This function draws a particular row values to see distribution"""
    plt.figure(figsize=(15,10))
    return sns.histplot(df[:][idx_row-1:idx_row].values.tolist())
```

```python
def plot_A_n(df,row,col_index):
    """This functions draws the values for specific rows"""
    while col_index <= 10:
        
        plt.plot(df[row-1:row][[col for col in df.columns if 'A'.join(str(col_index)) in col]].values.reshape(-1,1))
        col_index +=1
        plt.show()

#I wanted to see i can draw a plot of all the indices with A+(SOME NUMBER) for example
```

```python
df3 = df2[df2['target']=='Streptococcus_pyogenes']
plot_A_n(df3, 10,1)
```
