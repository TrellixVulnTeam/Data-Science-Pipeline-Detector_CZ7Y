```python
import numpy as np
import pandas as pd
```

```python
#Import the train data on a DataFrame

df = pd.read_csv("../input/tabular-playground-series-feb-2022/train.csv")
```

```python
#Take a peek at the data
df.head()
```

```python
#Check if there is any null element on the dataframe

df.isnull().sum().sum()
```

```python
#Drop the non features on the X and store target on y, then split the data on the training data and validation data

from sklearn.model_selection import train_test_split
X = df.drop(['target', 'row_id'], axis = 1)
y = df.target

X_train, X_valid, y_train, y_valid = train_test_split(X, y)
```

```python
#Encode the categorical data into int values

from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_valid_encoded = label_encoder.transform(y_valid)
```

```python
#Use a model and train it using training data 

from xgboost import XGBClassifier
model = XGBClassifier(eval_metric = 'merror')
model.fit(X_train, y_train_encoded)
```

```python
#Evalue the model

valid_predictions = model.predict(X_valid)
series_valid = pd.Series(valid_predictions)
accuracy = (series_valid == y_valid_encoded).sum()/series_valid.size
print(accuracy)
```

```python
#Load the data we want to use to predict and do the predictions, then format the data to fit the competition format

df_test = pd.read_csv("../input/tabular-playground-series-feb-2022/test.csv")
predictions = model.predict(df_test.drop(['row_id'], axis = 1))
prediction_df = pd.DataFrame(label_encoder.inverse_transform(predictions), columns = ['target']).set_index(df_test.row_id)
```

```python
#Create the file to submit to competition
prediction_df.to_csv("my_submission.csv")

#Next steps could be doing some feature engeneering, and maybe tune the model parameters.
```
