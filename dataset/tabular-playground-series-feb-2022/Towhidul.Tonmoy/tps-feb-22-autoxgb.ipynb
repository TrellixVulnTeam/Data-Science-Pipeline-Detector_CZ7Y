```python
from autoxgb import AutoXGB
```

```python
train_filename = "../input/tabular-playground-series-feb-2022/train.csv"
test_filename = "../input/tabular-playground-series-feb-2022/test.csv"
```

```python
pd.read_csv(train_filename )
```

```python
output = "autoxgb1"

# optional parameters
#task = None
idx = "row_id"
targets = ["target"]
#features = None
#categorical_features = None
use_gpu = False
num_folds = 5
seed = 42
num_trials = 50
time_limit = 60
fast = True

# Now its time to train the model!
axgb = AutoXGB(
    train_filename=train_filename,
    output=output,
    test_filename=test_filename,
    idx=idx,
    targets=targets,
    use_gpu=use_gpu,
    num_folds=num_folds,
    seed=seed,
    num_trials=num_trials,
    time_limit=time_limit,
    fast=fast,
)
axgb.train()
```
