## 1. Import the dataset and libraries

```python
# import the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import the dataset
train_dataset = pd.read_csv("../input/tabular-playground-series-feb-2022/train.csv", index_col = 0)
test_dataset = pd.read_csv("../input/tabular-playground-series-feb-2022/test.csv", index_col = 0)
```

### Dropping duplicates

```python
# checking for duplicates
print("train dataset shape (before dropping duplicates)", train_dataset.shape)
print("duplicated rows in train dataset: ", train_dataset.duplicated().sum())

# dropping duplicated rows
train_dataset.drop_duplicates(keep = 'first', inplace = True)
print("train dataset shape (after dropping duplicates)", train_dataset.shape)
print("duplicated rows in train dataset: ", train_dataset.duplicated().sum())
```
