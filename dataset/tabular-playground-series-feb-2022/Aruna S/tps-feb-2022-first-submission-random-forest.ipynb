# **SPLITTING THE FEATURES AND LABEL**

> **Target variable is the only dependent variable and hence taking it as the label (y) while taking rest of the variables as features (x)**

```python
#splitting dataset into features and label

x = train.drop(columns='target')
y = train['target']
```

```python
#displaying the features

x
```

```python
#displaying the label

y
```

# **TRAIN TEST SPLIT**

> **Let us split the dataset into train and test set such that 33% of the dataset is for training while the rest is for testing**

```python
#importing train test split from sklearn library

from sklearn.model_selection import train_test_split
```

```python
#train test split
    
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

# **RANDOM FOREST**

**IMPORTING RANDOM FOREST CLASSIFIER FROM SKLEARN LIBRARY**

```python
#importing library

from sklearn.ensemble import RandomForestClassifier
```

**MODEL BUILDING**

```python
#creating a random forest model

rf = RandomForestClassifier()
```

**FITTING THE MODEL**

```python
#fitting the model

rf.fit(x_train,y_train)
```

# **MODEL PREDICTIONS**

**MODEL PREDICTION**

> **Let us perform the prediction on our train set first**

```python
#model prediction on the train set

rf.predict(x_train)
```

```python
#model prediction on the test set

rf.predict(x_test)
```

```python
#displaying the initial sample file with default values

sample
```

> **Let us see how our model performs on the testing dataset**

```python
#predicting on the test class


rf.predict(test)
```

```python
#storing our test class predictions in a variable 

pred = rf.predict(test)
```

```python
#predictions copied to the submissions file

sample['target'] = pred
sample.to_csv('predictions.csv')
```
