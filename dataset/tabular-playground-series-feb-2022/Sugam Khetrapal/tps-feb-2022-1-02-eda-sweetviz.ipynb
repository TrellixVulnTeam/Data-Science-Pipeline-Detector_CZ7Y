# Import Libraries

```python
from pandas import read_csv
import sweetviz as sv
```

# Load Dataset

```python
# Load Dataset

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
        
filename_training          = '/kaggle/input/tabular-playground-series-feb-2022/train.csv'
dataset_training           = read_csv(filename_training)
```

# Print Report

```python
advert_report = sv.analyze([dataset_training, 'target'], pairwise_analysis='off')
```

```python
advert_report.show_html()
```
