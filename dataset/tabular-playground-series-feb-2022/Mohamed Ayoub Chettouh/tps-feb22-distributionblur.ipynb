# Data Engineering:

Blurring the distribution to improve resillience to mutations/Instrument Error.

```python
def mer10_arrtostr(mer10):
    assert type(mer10)==list, ValueError('Input must be a list.')
    assert len(mer10)==4, ValueError('Length of input array must be 4.')
    return f"A{mer10[0]}T{mer10[1]}G{mer10[2]}C{mer10[3]}"
def mer10_strtoarr(string):
    assert type(string)==str, ValueError('Input must be a string.')
    groups = re.search(r'A([0-9]+)T([0-9]+)G([0-9]+)C([0-9]+)', string).groups()
    return [int(groups[0]), int(groups[1]), int(groups[2]), int(groups[3])]

def adjacent_mers(mer10, string_type=False):
    adj_mers = []
    if type(mer10) == str:
        mer10 = mer10_strtoarr(mer10)
    for i in range(4):
        if mer10[i] == 0:
            continue
        mer10[i] -= 1
        for j in range(4):
            if i == j or mer10[j] == 10:
                continue
            mer10_added = mer10.copy()
            mer10_added[j] += 1
            adj_mers.append(mer10_added)
        mer10[i] += 1
    if not string_type:
        return adj_mers
    for i in range(len(adj_mers)):
        adj_mers[i] = mer10_arrtostr(adj_mers[i])
    return set(adj_mers)
```

```python
adjacency_dict = {}
for mer10 in data_range:
    adjacency_dict[mer10] = adjacent_mers(mer10 ,string_type=True)
```

```python
def blur_row(row, err=0.15):
    
    ret_row = pd.Series(0, index=row.index, dtype=float)
    ret_row['target'] = row['target']
    for mer10 in data_range:
        value = row[mer10]
        n_adj = len(adjacency_dict[mer10])
        err_per_mer10 = value * err / n_adj
        for adj_mer10 in adjacency_dict[mer10]:
            ret_row[adj_mer10] += err_per_mer10
        ret_row[mer10] += value * (1-err)
    ret_row['target']
    return ret_row
def blur_dataframe(df, err=0.15):
    ret_df = df.copy()
    #ret_df['target'] = df['target']
    for mer10 in data_range:
        ret_df[mer10] = (df[adjacency_dict[mer10]].sum(axis=1) * err) + df[mer10] * (1-err)
    return ret_df
```

Blurring coefficient:

```python
ERR = 0.25
df_train_blurred = blur_dataframe(df_train, err=ERR)
df_test_blurred = blur_dataframe(df_test, err=ERR)
```

# Training:

```python
from sklearn.preprocessing import LabelEncoder

# Encoding categorical features
le = LabelEncoder()

DATAFRAME = df_train_blurred
df_train_blurred['sample_weight'] = 1

X = DATAFRAME[data_range]
y = pd.DataFrame(le.fit_transform(DATAFRAME['target']), columns=['target'])

X_test = df_test_blurred[data_range]

sample_weight = DATAFRAME['sample_weight']
```

```python
from tqdm import tqdm

from sklearn.ensemble import ExtraTreesClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import StratifiedKFold

N_SPLITS = 10

folds = StratifiedKFold(n_splits=N_SPLITS, shuffle=True)
y_pred_list, y_proba_list, scores = [], [], []

valid_df = pd.DataFrame()
valid_df['true_target'] = DATAFRAME.target
valid_df['pred_target'] = DATAFRAME.target

n_estimators = 128
max_depth = 64
min_samples_split = 3
min_samples_leaf = 1
criterion = 'gini'

for fold, (train_id, valid_id) in enumerate(tqdm(folds.split(X, y), total=N_SPLITS)):
    print('####### Fold: ', fold)
    
    # Splitting
    X_train, y_train, sample_weight_train = X.iloc[train_id], y.iloc[train_id], sample_weight.iloc[train_id]
    X_valid, y_valid, sample_weight_valid = X.iloc[valid_id], y.iloc[valid_id], sample_weight.iloc[valid_id]
    
    # Model
    model = ExtraTreesClassifier(
        n_estimators=n_estimators,
        max_depth=max_depth,
        min_samples_split=min_samples_split,
        min_samples_leaf=min_samples_leaf,
        criterion=criterion,
        n_jobs=-1,
        verbose=0,
        random_state=fold
    )

    # Training
    model.fit(X_train, y_train, sample_weight_train)
        
    # Validation
    valid_pred = model.predict(X_valid)
    valid_score = accuracy_score(y_valid, valid_pred, sample_weight=sample_weight_valid)
    print(f'Accuracy score: {valid_score:5f}\n')
    scores.append(valid_score)
    
    #Adding Validated Data to the confusion dataframe
    valid_df.at[valid_id, 'pred_target'] = le.inverse_transform(valid_pred)
    
    # Prediction for submission
    y_pred_list.append(model.predict(X_test))
    y_proba_list.append(model.predict_proba(X_test))
    
score = np.array(scores).mean()
print(f'Mean accuracy score: {score:6f}')
```
