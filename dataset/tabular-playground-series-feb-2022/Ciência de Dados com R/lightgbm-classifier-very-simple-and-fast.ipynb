```python
df_train = pd.read_csv('/kaggle/input/tabular-playground-series-feb-2022/train.csv')
df_test = pd.read_csv('/kaggle/input/tabular-playground-series-feb-2022/test.csv')
```

```python
df_train.head(1)
```

```python
X = df_train.iloc[:,1:-1]
y = df_train.iloc[:,-1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
my_lgbm = lgbm.LGBMClassifier(max_depth=8,learning_rate=0.01,num_iterations=200)
```

```python
my_lgbm
```

```python
my_lgbm.fit(X_train,y_train)
```

```python
treino = my_lgbm.predict(X_train)
teste = my_lgbm.predict(X_test)
```

```python
print(accuracy_score(treino,y_train))
print(accuracy_score(teste,y_test))
```

```python
target = my_lgbm.predict(df_test.iloc[:,1:])
```

```python
submission = df_test.iloc[:,:1].copy()
```

```python
submission['target'] = target
```

```python
submission.to_csv('./submission.csv', index=False)
```
