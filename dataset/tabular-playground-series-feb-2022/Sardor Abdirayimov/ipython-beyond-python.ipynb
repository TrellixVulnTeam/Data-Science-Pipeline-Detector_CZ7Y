## `?` & `??`| Instead of `help()`

Human being are not ideal species.
In Programming ecosystem, it is really challenging to remember specific methods and function or to forget some theirs concepts.
For faster memorizing that functions or methods, IPython provides several 
In following example, let\`s image that we forget how `classification_report` works or what arguments it requires.

```python
from sklearn.metrics import classification_report
# open outpurt to see docstring(It is too long)
classification_report?
```

After reading previous doctring and required arguments for that model, We got answer for our question very quickly.
However, sometimes we need source code to look it to ununderstand actions behind specific method, function, class.
For that task, we have got `inspect` library. Follow me and get source code for example for `np.arange`

```python
import inspect 
import numpy as np 

inspect_version = inspect.getsource(np)
print(inspect_version) # open outpurt to see docstring(It is too long)
```

## Great symbol in IPython: `*`

As you noticed, `Tab` keyboard is really helpful, when we know nothing about  objects and attributes or we know first few letters of them. 
However, what to do if you remember middle of last characters of the object or attributes.
For that situations, IPython provides `*` symbol, which matches any string, including empty string. 
For example:
I forget `pct` letters of the following attribute `pandas.DataFrame.pct_change`.
I use `df.*change?` to find that forgotten letters

```python
import pandas as pd 
import warnings
warnings.filterwarnings("ignore")
df = pd.DataFrame(bac_families, index=np.arange(3))

df.*change?
```

## IPython magic commands

Every magic command has prefix of either `%` or `%%`.

`%` single symbol is used, when you want magic command to be implemented into a singe cell.

`%%` double symbol is used, when you want magic command to be implemented into a multiple cells

In order to see al magic commands:
write `%magic` command.

In order to read docstring of specific magic command, as i meantioned before use `?` symbol in the end.

I usually use nest commands in my notebooks: `%timeit`, `%run`, `%copy` and `%lprun`.
However, there are many other magic commads, which helps you to create kernel more effectively such as `%memit` and so on.



```python
%magic # open outpurt to see all magic commands(It is very long)
```

## `_` Underscore Shortcut

In IPython, you can use underscore(`_` symbol) to get access to previous output results. 

Moreover, you can use double underscore to get output results of second-to-last output result, even triple underscore to access to third-to-last output.

```python
bac_families.keys()
```

```python
bac_families.items()
```

```python
print(_) # get results of previous output
```

```python
print(__) # get results of second-to-last output
```
