# import libraries

```python
import numpy as np # linear algebra
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

# Load Data

```python
train=pd.read_csv('/kaggle/input/tabular-playground-series-feb-2022/train.csv')
test=pd.read_csv('/kaggle/input/tabular-playground-series-feb-2022/test.csv')
```

```python
train.isnull().values.any()
```

```python
test.drop('row_id',axis=1,inplace=True)
train.drop('row_id',axis=1,inplace=True)
```

```python
print(train[train.duplicated()])
```

```python
train.shape
```

```python
train.drop_duplicates(keep='first',inplace=True)
```

```python
train_target=train['target']
train.drop('target',axis=1,inplace=True)
```

# Check Predictions

```python
x_train,x_test,y_train,y_test=train_test_split(train,train_target,test_size=0.2,random_state=123)
```

```python
clf=KNeighborsClassifier(n_neighbors=1, p=2)
clf.fit(x_train,y_train)
pred=clf.predict(x_test)
```

```python
confusion_matrix(y_test,pred)
```

```python
accuracy_score(y_test,pred)
```

# Without removing Duplicates

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
sample_submission = pd.read_csv('../input/tabular-playground-series-feb-2022/sample_submission.csv')
test = pd.read_csv('../input/tabular-playground-series-feb-2022/test.csv')
train = pd.read_csv('../input/tabular-playground-series-feb-2022/train.csv')
features = [c for c in train.columns]
target = train['target']
features.pop(0)
features.pop(-1)

train = train[features]
test = test[features]

train = np.array(train)
target = np.array(target)
test = np.array(test)
clf = KNeighborsClassifier(n_neighbors=1, p=2)

clf.fit(train, target)
y_ = clf.predict(test)

sample_submission['target'] = y_
sample_submission.to_csv('output.csv', index=False)
```
