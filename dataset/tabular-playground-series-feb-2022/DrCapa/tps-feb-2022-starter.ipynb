# Libraries

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder
from xgboost import XGBClassifier
from sklearn.metrics import mean_squared_error, make_scorer
```

# Path


```python
path = '/kaggle/input/tabular-playground-series-feb-2022/'
os.listdir(path)
```

# Load Data

```python
train_data = pd.read_csv(path+'train.csv')
test_data = pd.read_csv(path+'test.csv')
samp_subm = pd.read_csv(path+'sample_submission.csv')
```

# Overview

```python
print('Number train samples:', len(train_data.index))
print('Number train features:', len(train_data.columns))
print('Number test samples:', len(test_data.index))
```

# EDA

```python
train_data['target'].value_counts()
```

# Model

Define train, validation and test data:

```python
X = train_data[train_data.columns[1:-1]]
y = train_data['target']

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = 0.2, random_state=2022)

X_test = test_data[test_data.columns[1:]]
```

Define and train the model:

```python
model = XGBClassifier()
model.fit(X_train, y_train, eval_set=[(X_val, y_val)], verbose=True)
```

Predict test data:

```python
y_test = model.predict(X_test)
```

# Export

```python
samp_subm['target']=y_test
```

```python
samp_subm['target'].value_counts()
```

```python
samp_subm.to_csv('submission.csv', index=False)
```
