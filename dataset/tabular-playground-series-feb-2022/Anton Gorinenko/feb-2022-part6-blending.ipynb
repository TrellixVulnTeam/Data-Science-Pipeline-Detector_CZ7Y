```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from scipy import stats
```

```python
train_df = pd.read_csv('../input/tabular-playground-series-feb-2022/train.csv', index_col="row_id")
test_df = pd.read_csv('../input/tabular-playground-series-feb-2022/test.csv')
```

```python
le = LabelEncoder()
le.fit(train_df.target)

preds = []
submits = [
    '../input/feb-2022-part2-cat-boost-classifier/submission.csv',
    '../input/feb-2022-part3-lgbm-classifier/submission.csv',
    '../input/feb-2022-part4-xgb-classifier/submission.csv',
    '../input/feb-2022-part5-random-forest-classifier/submission.csv'
          ]
for filename in submits:
    df = pd.read_csv(filename)
    preds.append(le.transform(df.target))

preds = np.array(preds)
print(preds.shape)
preds = stats.mode(preds, axis=0)[0].flatten()
```

# Submission

```python
def save_submission(y_pred):  
    y_pred = y_pred.astype(np.int64)
    y_pred_class = le.inverse_transform(y_pred)
    submission = test_df[['row_id']].copy() 
    submission["target"] = y_pred_class
    
    assert len(y_pred_class) == submission.shape[0]
    assert 2 == submission.shape[1]
    
    submission.to_csv("submission.csv",index=False)
    return submission
```

```python
save_submission(preds.flatten()).head()
```
