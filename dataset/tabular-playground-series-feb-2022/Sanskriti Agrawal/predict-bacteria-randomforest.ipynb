```python
train_data = pd.read_csv('/kaggle/input/tabular-playground-series-feb-2022/train.csv')
train_data.head(6)
```

```python
test_data = pd.read_csv('/kaggle/input/tabular-playground-series-feb-2022/test.csv')
test_data.shape
```

```python
X_train = train_data.iloc[:,1:287]
X_test = test_data.iloc[:,1:287]
y_train = train_data.iloc[:,-1]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```python
from sklearn.model_selection import train_test_split
X_train, Xx_test, y_train, yy_test = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

```python
model1 = RandomForestClassifier(n_estimators=100)
```

```python
model1.fit(X_train,y_train)
```

```python
prediction11 = model1.predict(Xx_test)


print(classification_report(yy_test, prediction11))
```

```python
predicts = model1.predict(X_test)
```

```python
output = pd.DataFrame({'row_id': test_data.row_id, 'target': predicts})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```
