```python
kf = model_selection.KFold(n_splits=7, shuffle=True, random_state=42)
for fold, (train_indicies, valid_indicies) in enumerate(kf.split(X=train)):
    train.loc[valid_indicies, "kfold"] = fold
```

```python
train
```

```python
train["kfold"].hist()
```

```python
import matplotlib.pyplot as plt
a=0
r, c = 2, 4
fig = plt.figure(figsize=(8*c, 6*r))
for _r in range(r):
    for _c in range(c):
        plt.subplot(r, c, _r*c + _c + 1)
        if a==7:
            plt.axis(False)
            break
        else:
            df=train[train["kfold"]==a]
            df["target"].hist(bins=20,xlabelsize=8,xrot=45)
            label = f"kflod{a}"
            plt.title(label)
            #plt.axis(False)
            a+=1
plt.show()
```
