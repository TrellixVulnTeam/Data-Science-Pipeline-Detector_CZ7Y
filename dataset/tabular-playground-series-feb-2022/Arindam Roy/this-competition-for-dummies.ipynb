```python
import pandas as pd
import numpy as np
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler , LabelEncoder, MinMaxScaler
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from scipy.stats import mode
```

```python
input_df = pd.read_csv('../input/tabular-playground-series-feb-2022/train.csv')
input_df.drop_duplicates(inplace=True)
# enc = LabelEncoder()
# input_df['target'] = enc.fit_transform(np.array(input_df['target']).reshape(-1, 1))
# (Dont really need to encode in a tree based ensemble)
```

```python
# Straight into it :X
X = input_df.drop('target',axis=1)
y = input_df['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# Skipping converting to integer
# Skipping GCD column
# Check my other notebook for that!

from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
clf = RandomForestClassifier(max_depth=700, random_state=0, criterion='gini')
```

```python
clf.fit(X_train, y_train)
pred = clf.predict(X_test)
print(f'Acuracy of the random forest: {accuracy_score(y_test, pred)}')
```

```python
# Getting the submission done

final_df = pd.read_csv('../input/tabular-playground-series-feb-2022/test.csv')
pred_final = clf.predict(final_df)
output_csv = pd.DataFrame()
output_csv['target'] = pred_final
output_csv['row_id'] = final_df.row_id
output_csv.to_csv('submission_joke_rf.csv',index=False)
```

```python
# Now to the extra trees
from sklearn.ensemble import ExtraTreesClassifier
clf_et = ExtraTreesClassifier(n_estimators=700, random_state=0)
clf_et.fit(X_train, y_train)
pred = clf_et.predict(X_test)
print(f'Acuracy of the extra trees: {accuracy_score(y_test, pred)}')
```

```python
# Getting the submission done
final_df = pd.read_csv('../input/tabular-playground-series-feb-2022/test.csv')
pred_final = clf_et.predict(final_df)
output_csv = pd.DataFrame()
output_csv['target'] = pred_final
output_csv['row_id'] = final_df.row_id
output_csv.to_csv('submission_joke_xt.csv',index=False)
```
