```python
df_test = pd.read_csv('../input/tabular-playground-series-feb-2022/test.csv')
df_test = df_test.set_index('row_id')
df_test
```

```python
df_test.isnull().sum().sum()
```

```python
from supervised.automl import AutoML
```

```python
X = df_train.drop('target', axis = 1)
y = df_train['target']
```

```python
# load the data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42, shuffle=y)
```

```python
# train models with AutoML
automl = AutoML(mode="Optuna", algorithms=["Random Forest", "Extra Trees", 'LightGBM', 'Neural Network'], eval_metric = 'logloss')
automl.fit(X_train, y_train)

# compute the accuracy on test data
predictions = automl.predict_all(X_test)
print(predictions.head())
```

```python
automl.predict_all(df_test)
```

```python
pred = automl.predict(df_test)
```

```python
subb = pd.read_csv('../input/tabular-playground-series-feb-2022/sample_submission.csv')
subb
```

```python
subb.target = pred
```
