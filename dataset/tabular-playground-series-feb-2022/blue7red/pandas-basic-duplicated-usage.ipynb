```python
import numpy as np
import pandas as pd 
```

```python
TRAIN_PATH = "../input/tabular-playground-series-feb-2022/train.csv"
ID = "row_id"
TARGET = "target"
```

```python
train = pd.read_csv(TRAIN_PATH)
```

# get feature names

```python
feature = [col for col in train.columns if col != ID and col != TARGET]
print(feature)
```

# get duplicated size 

```python
duplicated_train = train[train[feature].duplicated()==True]
print("duplicated size = ",len(duplicated_train))
```

# get unique train dataframe

```python
print("original train items length",len(train))
train = train[train[feature].duplicated()==False]
print("unique train items length",len(train))
```
