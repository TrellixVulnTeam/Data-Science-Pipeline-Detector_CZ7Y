```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

# Introduction

Purpose of this workbook is to test some easy, beginner level models and select the one which gives the best results. 

For all models I am using default parameters. With a proper tune-up of parameters we would get better results.

# Load Data

```python
%%time
train = pd.read_csv('../input/tabular-playground-series-feb-2022/train.csv')
test = pd.read_csv('../input/tabular-playground-series-feb-2022/test.csv')
sample = pd.read_csv('../input/tabular-playground-series-feb-2022/sample_submission.csv')

#display(train,test, sample)
```

```python
X = train.drop(['row_id','target'], axis = 1)
y = train['target']
X_train, X_test, y_train, y_test = train_test_split(X,y)

#display(X_train, X_test, y_train, y_test)
```

# Predict

## One Label for all

This is just a dummy algorithm. The purpose is to see if I can generate submission in correct format. No logic inside, just randomly selected label

```python
pred = ['Enterococcus_hirae' for i in range(len(y_test))] # randomly selected label

# score
accuracy_score(y_test, pred)
```

## XGBClasifier model

```python
%%time
from xgboost import XGBClassifier

model_xgb = XGBClassifier(random_state=2)
model_xgb.fit(X_train,y_train)
```

```python
%%time
pred = model_xgb.predict(X_test)

display(pred)

# score
accuracy_score(y_test, pred)
```

## Light GBM

```python
%%time
import lightgbm as ltb

model_ltb = ltb.LGBMClassifier(random_state=2)
model_ltb.fit(X_train, y_train)
```

```python
%%time
pred = model_ltb.predict(X_test)

display(pred)

# score
accuracy_score(y_test, pred)
```

# Generate Submission File

XGBClasifier performed slightly better - lets use it for submission.

We will fit the model once more, this time we use the whole dataset.

```python
model_xgb.fit(X, y)
pred = model_xgb.predict(test.drop('row_id', axis = 1))
```

```python
submission = pd.DataFrame(test['row_id'])
submission['target'] = pred

display(submission)

submission.to_csv('submission.csv', index=False)
```
