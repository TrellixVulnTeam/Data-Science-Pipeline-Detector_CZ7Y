import useful libraries for python.

```python
import numpy as np
import pandas as pd
```

then, import csv files.

```python
train = pd.read_csv('../input/tabular-playground-series-feb-2022/train.csv')
test = pd.read_csv('../input/tabular-playground-series-feb-2022/test.csv')
sample_submission = pd.read_csv('../input/tabular-playground-series-feb-2022/sample_submission.csv')
```

check how they look.

```python
train.head()
```

concatenate two files so that they are edited simultaneously.

```python
data = pd.concat([train, test], sort=False)
```

check the missing value.

```python
data.isnull().sum()
```

now, divide it.

```python
train = data[:len(train)]
test = data[len(train):]
```

```python
y_train = train['target']
X_train = train.drop('target', axis=1)
X_test = test.drop('target', axis=1)
```

this time, use random forest.

```python
from sklearn.ensemble import RandomForestClassifier


clf = RandomForestClassifier(n_estimators=100, max_depth=2, random_state=0)
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

create csv for submission.

```python
sub = sample_submission
sub['target'] = list(map(str, y_pred))
sub.to_csv("submission.csv", index=False)
```
