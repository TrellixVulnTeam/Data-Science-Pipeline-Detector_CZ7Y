```python
train = pd.read_csv('../input/tabular-playground-series-feb-2022/train.csv')
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder().fit(train.target)
train.target = le.transform(train.target)

y = train.target
elements = [c for c in train.columns if c not in ['row_id', 'target','gcd']]
```

## Remove Duplicates

```python
train = train.drop_duplicates(subset=elements)
train.shape
```

# Bias and GCD
This section is copied and inspired by [ambrosm](https://www.kaggle.com/ambrosm) and his notebook [here](https://www.kaggle.com/ambrosm/tpsfeb22-01-eda-which-makes-sense).


From the research paper attached, he pointed out the values were most probably integers with some bias error in all entries

```python

from math import factorial
def bias(w, x, y, z):
    return factorial(10) / (factorial(w) * factorial(x) * factorial(y) * factorial(z) * 4**10)

def bias_of(s):
    w = int(s[1:s.index('T')])
    x = int(s[s.index('T')+1:s.index('G')])
    y = int(s[s.index('G')+1:s.index('C')])
    z = int(s[s.index('C')+1:])
    return bias(w, x, y, z)
train_i = pd.DataFrame({col: ((train[col] + bias_of(col)) * 1000000).round().astype(int) for col in elements})
train_i
```

```python
def gcd_of_all(df_i):
    gcd = df_i[elements[0]]
    for col in elements[1:]:
        gcd = np.gcd(gcd, df_i[col])
    return gcd

train['gcd'] = gcd_of_all(train_i)

print('gcd values with frequncy : ',np.unique(train['gcd'], return_counts=True))
```

```python
from sklearn.decomposition import PCA, SparsePCA
from sklearn.decomposition import TruncatedSVD
from umap.umap_ import UMAP
from sklearn.random_projection import SparseRandomProjection
from sklearn.decomposition import KernelPCA
from sklearn.manifold import Isomap, LocallyLinearEmbedding, MDS
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python

def plot_components_gcd(train = None, train_i=None, pca = None, cmap=cmap, s=1):
    for scale in np.sort(train['gcd'].unique()):
        # Compute the PCA
        Xt_tr = pca.fit_transform(train_i[elements][train['gcd'] == scale], train.target[train['gcd'] == scale])
        
        # Plot a scattergram, projected to two PCA components, colored by classification target
        plt.figure(figsize=(6,6))
        plt.scatter(Xt_tr[:,0], Xt_tr[:,1], c=train.target[train['gcd'] == scale], s=s, cmap=cmap)
        plt.title(f"{1000000 // scale} decamers ({(train['gcd'] == scale).sum()} samples with gcd = {scale})")
        plt.show()
```

# Principal Component Analysis

## PCA

```python
plot_components_gcd(train, train_i, PCA())
```

## Truncated SVD

If you look at the data points closely, you'll find that that most of the entries are zero, hence the data is **sparse** in nature. `TruncatedSVD` works better with sparse data. 

```python
train_i.head().style.background_gradient(cmap=pal[5])
```

```python
plot_components_gcd(train, train_i, TruncatedSVD(),cmap=pal[0])
```

## SparsePCA
Finds the set of sparse components that can optimally reconstruct the data. The amount of sparseness is controllable by the coefficient of the L1 penalty, given by the parameter alpha.

**Read more [here](https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.SparsePCA.html)**





```python
plot_components_gcd(train, train_i, SparsePCA(n_components=2),cmap=pal[2])
```

It can be seen that data points with **lower gcd** can be easily distinguishable while, as the gcd increases, it becomes more and more convoluted and complex.

# KernelPCA
KernelPCA is an extension of PCA which achieves non-linear dimensionality reduction through the use of kernels 

## KerelPCA with `kernel='poly'`

```python
plot_components_gcd(train[:1000], train_i[:1000],
                    KernelPCA(kernel='poly'),
                    cmap='viridis', s=10)
```

## KerelPCA with `kernel='cosine'`

```python
plot_components_gcd(train[:1000], train_i[:1000],
                    KernelPCA(kernel='cosine'),
                    cmap=pal[9], s=10)
```

# UMAP
UMAP is a dimension reduction technique that can be used for visualisation similarly to t-SNE, but also for general non-linear dimension reduction.

**Read more [here](https://umap-learn.readthedocs.io/en/latest/)**

```python
plot_components_gcd(train, train_i, UMAP(),cmap=pal[1])
```

## SparseRandomProjections

```python

plot_components_gcd(train, train_i,
                    SparseRandomProjection(n_components=2),
                    cmap='jet')
```

## LinearDiscriminantAnalysis
while being a prediction model, it can also be used in Dimensionality Reduction. It is done so by projecting the input data to a linear subspace consisting of the directions which maximize the separation between classes 

```python
plot_components_gcd(train, train_i, LinearDiscriminantAnalysis(), cmap=pal[10])
```

# Manifold Learning

Manifold Hypotheses assumes that most of the higher dimensional lies close to a much lower-dimension manifold
image.png

Manifold Learning Algorithms requires a loooot of memory to compute, which is not feasibe for the current dataset. Hence, we will be using the first 1000 data points for computation

## Isomap
short for *Isometric Mapping* ->  Isomap seeks a lower-dimensional embedding which maintains geodesic distances between all points.

```python
plot_components_gcd(train[:1000], train_i[:1000],
                    Isomap(),
                    cmap=pal[4],s=10)
```

## LocallyLinearEmbedding
Locally linear embedding (LLE) seeks a lower-dimensional projection of the data which preserves distances within local neighborhoods

```python
plot_components_gcd(train[:1000], train_i[:1000],
                    LocallyLinearEmbedding(n_components=2, eigen_solver='dense'),
                    cmap=pal[8], s=10)
```

## MDS
Multi Dimensional Scaling (MDS) seeks a low-dimensional representation of the data in which the distances respect well the distances in the original high-dimensional space.

```python
plot_components_gcd(train.iloc[:1000], train_i[:1000],
                   MDS(),
                   cmap=pal[6], s=10)
```
