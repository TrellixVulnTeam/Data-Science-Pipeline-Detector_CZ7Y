```python
train_df['target'] = train_df['target'].replace(['Bacteroides_fragilis'],0)            #replacing 'Bacteroides_fragilis' with 0
train_df['target'] = train_df['target'].replace(['Streptococcus_pyogenes'],1)          #replacing 'Streptococcus_pyogenes' with 1
train_df['target'] = train_df['target'].replace(['Streptococcus_pneumoniae'],2)        #replacing 'Streptococcus_pneumoniae' with 2
train_df['target'] = train_df['target'].replace(['Campylobacter_jejuni'],3)            #replacing 'Campylobacter_jejuni' with 3
train_df['target'] = train_df['target'].replace(['Salmonella_enterica'],4)             #replacing 'Salmonella_enterica' with 4
train_df['target'] = train_df['target'].replace(['Escherichia_coli'],5)                #replacing 'Escherichia_coli' with 5
train_df['target'] = train_df['target'].replace(['Enterococcus_hirae'],6)              #replacing 'Enterococcus_hirae' with 6
train_df['target'] = train_df['target'].replace(['Escherichia_fergusonii'],7)          #replacing 'Escherichia_fergusonii' with 7
train_df['target'] = train_df['target'].replace(['Staphylococcus_aureus'],8)           #replacing 'Staphylococcus_aureus' with 8
train_df['target'] = train_df['target'].replace(['Klebsiella_pneumoniae'],9)           #replacing 'Klebsiella_pneumoniae' with 9
```

```python
train_df['target'].value_counts()
```

```python
y = train_df['target']
X = train_df.copy()
X.drop(['row_id', 'target'], inplace=True, axis=1)
print('X Shape: ', X.shape)
print('Y Shape: ', y.shape)
```

```python
Xtest = test_df.copy()
Xtest.drop('row_id', axis=1, inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
PredictorScaler = StandardScaler()
TestPredictorScaler = StandardScaler()

# Storing the fit object for later reference
PredictorScalerFit = PredictorScaler.fit(X)
TestPredictorScalerFit = TestPredictorScaler.fit(Xtest)
 
# Generating the standardized values of X
X = PredictorScalerFit.transform(X)
Xtest = TestPredictorScalerFit.transform(Xtest)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=27)
```

```python
X_train.shape
```

```python
# importing the libraries
from keras.models import Sequential
from keras.layers import Dense
 
# create ANN model
model = Sequential()
 
# Defining the Input layer and FIRST hidden layer, both are same!
model.add(Dense(units=35, input_dim=286, kernel_initializer='normal', activation='sigmoid'))
 
# Defining the Second layer of the model
# after the first layer we don't have to specify input_dim as keras configure it automatically
model.add(Dense(units=27, kernel_initializer='normal', activation='sigmoid'))

# Defining the Third layer of the model
model.add(Dense(units=19, kernel_initializer='normal', activation='sigmoid'))

# Defining the Fourth layer of the model
model.add(Dense(units=15, kernel_initializer='normal', activation='sigmoid'))

# Defining the Fifth layer of the model
model.add(Dense(units=11, kernel_initializer='normal', activation='sigmoid'))

# Defining the Sixth layer of the model
model.add(Dense(units=9, kernel_initializer='normal', activation='sigmoid'))

# Defining the Seventh layer of the model
model.add(Dense(units=7, kernel_initializer='normal', activation='sigmoid'))

# Defining the Eighth layer of the model
model.add(Dense(units=5, kernel_initializer='normal', activation='sigmoid'))

# Defining the Ninth layer of the model
model.add(Dense(units=3, kernel_initializer='normal', activation='sigmoid'))

# The output neuron is a single fully connected node 
# Since we will be predicting a single number
model.add(Dense(1, kernel_initializer='normal', activation='sigmoid'))
 
# Compiling the model
model.compile(loss='categorical_crossentropy', optimizer='rmsprop', metrics=['accuracy'])
 
# Fitting the ANN to the Training set
model.fit(X_train, y_train ,batch_size = 20, epochs = 10, verbose=1)
```

```python
y_pred_test = model.predict(Xtest)

df_result_1 = pd.DataFrame(y_pred_test, columns=['target'])

row_id = test_df[['row_id']]

df_result_2 = pd.concat([row_id.reset_index(drop=True), df_result_1.reset_index(drop=True)], axis=1)
```
