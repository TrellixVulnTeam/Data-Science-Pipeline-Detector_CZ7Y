```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', 10)

from pathlib import Path
TEMP_DIR = Path("/kaggle/temp/")
INPUT_DIR = next(Path("/kaggle/input/").iterdir())
```

```python
train = pd.read_csv(INPUT_DIR / "train.csv", index_col="row_id")
test = pd.read_csv(INPUT_DIR / "test.csv", index_col="row_id")
```

Let's look at the intersection of the rows between the training and test set.

```python
pd.merge(train, test, on=list(test.columns), how="inner")
```

**There are 1521 rows in common! No excuse to get a score less than 0.01521, haha!**
