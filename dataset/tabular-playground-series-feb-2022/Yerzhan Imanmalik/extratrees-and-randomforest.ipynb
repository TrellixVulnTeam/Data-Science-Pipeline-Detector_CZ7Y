```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
from tqdm import tqdm

from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
train_df = pd.read_csv('/kaggle/input/tabular-playground-series-feb-2022/train.csv')
test_df = pd.read_csv('/kaggle/input/tabular-playground-series-feb-2022/test.csv')
submission = pd.read_csv('/kaggle/input/tabular-playground-series-feb-2022/sample_submission.csv')
```

```python
features = train_df.columns[1:-1]
```

## ExtraTrees (0.972)

```python
lb = LabelEncoder()
X = train_df[features]
y = lb.fit_transform(train_df['target'])
lr = ExtraTreesClassifier(n_estimators=400)
lr.fit(X, y)
```

```python
preds = lr.predict(test_df.iloc[:, 1:])
submission.target = lb.inverse_transform(preds)

submission.to_csv("submission_extra_trees.csv", index=False)
```

## RandomForest (0.970)

```python
rf = RandomForestClassifier(n_estimators=400)
rf.fit(X, y)
```

```python
preds = rf.predict(test_df.iloc[:, 1:])
submission.target = lb.inverse_transform(preds)

submission.to_csv("submission_random_forest.csv", index=False)
```
