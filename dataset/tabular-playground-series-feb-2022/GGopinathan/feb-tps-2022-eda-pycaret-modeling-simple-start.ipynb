```python
data = train_df_all.sample(frac=0.9, random_state=42)
data_unseen = train_df_all.drop(data.index)
data.reset_index(inplace=True, drop=True)
data_unseen.reset_index(inplace=True, drop=True)
```

```python
from pycaret.classification import *
```

```python
clf01 = setup(data=data, pca = True, use_gpu = True, data_split_stratify= True, target='target', silent=True, session_id=123)
```

```python
# best = compare_models(include = ['rf','dt','knn','et'])
```

```python
clf_et = create_model('et')
```

```python
plot_model(clf_et, plot='auc')
```

```python
plot_model(clf_et, plot='confusion_matrix', plot_kwargs={'percent':True})
```

```python
plot_model(clf_et, plot='feature')
```

```python
final_et = finalize_model(clf_et)
```

```python
unseen_preds = predict_model(final_et, data=data_unseen)
from pycaret.utils import check_metric
check_metric(unseen_preds['target'], unseen_preds['Label'], metric='Accuracy')
```

```python
test_preds = predict_model(final_et, data=test_df)
test_preds.head()
```

```python
test_preds['Label']
```

```python
test_preds = test_preds.rename(columns={'Label':'target'})
```

```python
submission_df = pd.read_csv('/kaggle/input/tabular-playground-series-feb-2022/sample_submission.csv', header=0)
```

```python
submission_df.set_index('row_id', inplace=True)
submission_df = submission_df.assign(target= test_preds['target'])
```

```python
submission_df.reset_index(drop=False, inplace=True)
```
