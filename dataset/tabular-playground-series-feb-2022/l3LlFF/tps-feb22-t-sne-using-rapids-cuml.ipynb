```python
import cudf
from cuml.manifold import TSNE
import pandas as pd
import plotly.express as px
```

```python
%%time
train = cudf.read_csv('../input/tabular-playground-series-feb-2022/train.csv', index_col=0)
```

```python
y = train['target']
train.drop(['target'], axis=1, inplace=True)
train_sample = train.sample(20000)
```

```python
%%time 
tsne    = TSNE(n_components=2)
tsne_2D = tsne.fit_transform(train_sample)
```

```python
df_temp = pd.DataFrame(tsne_2D.as_matrix(), columns=['x1', 'x2'])
df_temp['category'] = y[train_sample.index].to_array()
fig = px.scatter(df_temp, x="x1", y="x2", color="category")
fig.show()
```
