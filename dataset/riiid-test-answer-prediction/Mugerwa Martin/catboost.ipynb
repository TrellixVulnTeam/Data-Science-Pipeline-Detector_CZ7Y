```python
# Used most of coding from this kernel https://www.kaggle.com/lgreig/simple-lgbm-baseline
import riiideducation
import dask.dataframe as dd
import  pandas as pd
import numpy as np
from sklearn.metrics import roc_auc_score 
from scipy import stats
import gc
import cudf
import itertools

%time
env = riiideducation.make_env()
train= cudf.read_csv('/kaggle/input/riiid-test-answer-prediction/train.csv',
                usecols=[1, 2, 3,4,6,7,8,9], dtype={'timestamp': 'int64', 'user_id': 'int32' ,'content_id': 'int16','content_type_id': 'int8','user_answer': 'int8','answered_correctly':'int8','prior_question_elapsed_time': 'float32','prior_question_had_explanation': 'boolean'}
              )
train = train[train.content_type_id == False]
#arrange by timestamp

def mode(x):
    mode = stats.mode(x)
    return mode[1]

def cat(x):
    cat = np.mean(x)
    return np.around(cat,decimals=1)

def count(x):
    count=x.count()
    return count

def bias(x):
    bias = mode(x)/count(x)
    return bias

train = train[train['answered_correctly']>=0]

train = train.sort_values(['timestamp'], ascending=True)

train.drop(['timestamp','content_type_id'], axis=1,   inplace=False)

train_user = cudf.DataFrame(columns=train.columns)

a=train[train['user_id']<1072000000]
b=train[train['user_id']>1072000000]

c=a[a['user_id']<536000000]
d=a[a['user_id']>536000000]

e = b[b['user_id']<1618000000]
f = b[b['user_id']>1618000000]

#---------------------User_ID_groups-------------------------------------#

g=c[c['user_id']<272000000].groupby('user_id')
h=c[c['user_id']>272000000].groupby('user_id')

i=d[d['user_id']<807000000].groupby('user_id')
j=d[d['user_id']>807000000].groupby('user_id')

k=e[e['user_id']<1346000000].groupby('user_id')
l=e[e['user_id']>1346000000].groupby('user_id')

m=f[f['user_id']<1885000000].groupby('user_id')
n=f[f['user_id']>1885000000].groupby('user_id')

#-----------------------------------------------------------------------#

#---------------------------------------------------------------------------------Free memory----------------------------------------# 
del train
del a
del b
del c
del d
del e
del f 
gc.collect()

#-------------------------------------------------------------------------------------------------------------------------------------#
for (user,df_user),(user1,df_user1),(user2,df_user2),(user3,df_user3),(user4,df_user4),(user5,df_user5),(user6,df_user6),(user7,df_user7) in itertools.zip_longest(g,h,i,j,k,l,m,n) :
    
    df_user['mean_user']=df_user['answered_correctly'].to_pandas().expanding(1).mean()
    df_user['sum_user']=df_user['answered_correctly'].to_pandas().expanding(1).sum()
    df_user['std_user']=df_user['answered_correctly'].to_pandas().expanding(1).std()
    df_user['skew_user']=df_user['answered_correctly'].to_pandas().expanding(3).skew()
    df_user['count_user']=df_user['answered_correctly'].to_pandas().expanding(1).count()
    
    df_user1['mean_user']=df_user1['answered_correctly'].to_pandas().expanding(1).mean()
    df_user1['sum_user']=df_user1['answered_correctly'].to_pandas().expanding(1).sum()
    df_user1['std_user']=df_user1['answered_correctly'].to_pandas().expanding(1).std()
    df_user1['skew_user']=df_user1['answered_correctly'].to_pandas().expanding(3).skew()
    df_user1['count_user']=df_user1['answered_correctly'].to_pandas().expanding(1).count()
    
    df_user2['mean_user']=df_user2['answered_correctly'].to_pandas().expanding(1).mean()
    df_user2['sum_user']=df_user2['answered_correctly'].to_pandas().expanding(1).sum()
    df_user2['std_user']=df_user2['answered_correctly'].to_pandas().expanding(1).std()
    df_user2['skew_user']=df_user2['answered_correctly'].to_pandas().expanding(3).skew()
    df_user2['count_user']=df_user2['answered_correctly'].to_pandas().expanding(1).count()
    
    df_user3['mean_user']=df_user3['answered_correctly'].to_pandas().expanding(1).mean()
    df_user3['sum_user']=df_user3['answered_correctly'].to_pandas().expanding(1).sum()
    df_user3['std_user']=df_user3['answered_correctly'].to_pandas().expanding(1).std()
    df_user3['count_user']=df_user3['answered_correctly'].to_pandas().expanding(1).count()
    
        
    df_user4['mean_user']=df_user4['answered_correctly'].to_pandas().expanding(1).mean()
    df_user4['sum_user']=df_user4['answered_correctly'].to_pandas().expanding(1).sum()
    df_user4['std_user']=df_user4['answered_correctly'].to_pandas().expanding(1).std()
    df_user4['skew_user']=df_user4['answered_correctly'].to_pandas().expanding(3).skew()
    df_user4['skew_user']=df_user4['answered_correctly'].to_pandas().expanding(1).count()
    
    df_user5['mean_user']=df_user5['answered_correctly'].to_pandas().expanding(1).mean()
    df_user5['sum_user']=df_user5['answered_correctly'].to_pandas().expanding(1).sum()
    df_user5['std_user']=df_user5['answered_correctly'].to_pandas().expanding(1).std()
    df_user5['skew_user']=df_user5['answered_correctly'].to_pandas().expanding(3).skew()
    df_user5['skew_user']=df_user5['answered_correctly'].to_pandas().expanding(1).count()
    
    df_user6['mean_user']=df_user6['answered_correctly'].to_pandas().expanding(1).mean()
    df_user6['sum_user']=df_user6['answered_correctly'].to_pandas().expanding(1).sum()
    df_user6['std_user']=df_user6['answered_correctly'].to_pandas().expanding(1).std()
    df_user6['skew_user']=df_user6['answered_correctly'].to_pandas().expanding(3).skew()
    df_user6['count_user']=df_user6['answered_correctly'].to_pandas().expanding(1).count()
    
    df_user7['mean_user']=df_user7['answered_correctly'].to_pandas().expanding(1).mean()
    df_user7['sum_user']=df_user7['answered_correctly'].to_pandas().expanding(1).sum()
    df_user7['std_user']=df_user7['answered_correctly'].to_pandas().expanding(1).std()
    df_user7['skew_user']=df_user7['answered_correctly'].to_pandas().expanding(3).skew()
    df_user7['count_user']=df_user7['answered_correctly'].to_pandas().expanding(1).count()
    
    train_user=train_user.append(df_user,ignore_index=True)
    train_user=train_user.append(df_user1,ignore_index=True)
    train_user=train_user.append(df_user2,ignore_index=True)
    train_user=train_user.append(df_user3,ignore_index=True)
    train_user=train_user.append(df_user4,ignore_index=True)
    train_user=train_user.append(df_user5,ignore_index=True)
    train_user=train_user.append(df_user6,ignore_index=True)
    train_user=train_user.append(df_user7,ignore_index=True)
    
    print(train_user)
    del df_user
    del df_user1
    del df_user2
    del df_user3
    del df_user4
    del df_user5
    del df_user6
    del df_user7
    gc.collect()
    
    if train_user['user_id'].count()>60000000:
        break
```
