```python
import pandas as pd
import numpy as np

train = pd.read_csv('/kaggle/input/riiid-test-answer-prediction/train.csv',
                   dtype={'row_id': 'int64',
                          'timestamp': 'int64',
                          'user_id': 'int32',
                          'content_id': 'int16',
                          'content_type_id': 'int8',
                          'task_container_id': 'int16',
                          'user_answer': 'int8',
                          'answered_correctly':'int8',
                          'prior_question_elapsed_time': 'float32',
                          'prior_question_had_explanation': 'boolean'}
                   )
```

```python
prior_question_elapsed_time = train.prior_question_elapsed_time.dropna()
prior_question_elapsed_time.mean()
```

Is this result correct?
### **No, the actual average is twice as high as this result!　工ｴｴｪｪ Σ”(⚙♊⚙ﾉ)ﾉ ｪｪｴｴ工‼

```python
total = np.zeros(1,dtype=np.float128)
for v in prior_question_elapsed_time:
    total[0] += v
print(total[0]/len(prior_question_elapsed_time))
```

**prior_question_elapsed_time.mean() is wrong!!!**

```python
prior_question_elapsed_time
```

```python
prior_question_elapsed_time.min(), prior_question_elapsed_time.max()
```

Even min and max are ok with float32, pandas series mean() returned wrong. 
If we use float64, it looks OK.

```python
prior_question_elapsed_time.astype('float64').mean()
```

numpy.mean() looks OK too even with float32.

```python
prior_question_elapsed_time.values
```

```python
prior_question_elapsed_time.values.mean()
```

I don't know the detail, but this problem happens baddly especially when the data is large.
May be it happens when the `sum` overflows flot32.

```python
for i in range(9):
    target_data = prior_question_elapsed_time[:10**i]
    diff = target_data.astype('float64').mean() - target_data.mean()
    print(f'{i} {abs(diff):.4f}')
```

So we need to be carefull with pandas mean(). The results of this experiment are:
<pre>
.mean() with float32 is NG
.mean() with float64 is OK
.values.mean() OK (you need to dropna beforehand for np.mean)
</pre>
