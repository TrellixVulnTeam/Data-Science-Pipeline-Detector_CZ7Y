Here was the function I used to calculate time since last action for the test data. It proved to be a decent feature (best one I have currently other than the obvious user accuracy history and question accuracy history). I am sure there is a better one out there that captures a similar idea, but this function could also be applied to other things regarding users' "state." I will warn this was computationally expensive and took 4-5 hours to submit.

Also, just for reference last_record is a dataframe with just the user_id and last timestamp for each user.
