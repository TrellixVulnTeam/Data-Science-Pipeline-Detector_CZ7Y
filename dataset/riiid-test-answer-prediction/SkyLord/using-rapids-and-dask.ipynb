```python
from dask.distributed import Client, wait

from dask.utils import parse_bytes
import cudf
import dask_cudf
```

```python
from dask_cuda import LocalCUDACluster
```

```python
!nvidia-smi
```

```python
cluster = LocalCUDACluster(
    CUDA_VISIBLE_DEVICES="0",
    rmm_pool_size=parse_bytes("14GB"), # This GPU has 16GB of memory
    device_memory_limit=parse_bytes("8GB"),
)
client = Client(cluster)
client
```

```python
%%time
train_ddf = dask_cudf.read_csv("/kaggle/input/riiid-test-answer-prediction/train.csv", chunksize="500 MB")
print(train_ddf.npartitions)
len(train_ddf)
```
