```python
full_train = pd.read_csv("/kaggle/input/riiid-test-answer-prediction/train.csv",nrows=20000)
questions = pd.read_csv('/kaggle/input/riiid-test-answer-prediction/questions.csv')
lectures = pd.read_csv('/kaggle/input/riiid-test-answer-prediction/lectures.csv')
```

```python
full_train = full_train.dropna()
```

```python
full_train.head()
```

```python
full_train = full_train[['row_id','user_id','content_id','content_type_id','answered_correctly']]
```

```python
feature_col = ['row_id','user_id','content_id','content_type_id']
target_col = ['answered_correctly']
```

```python
X = full_train[feature_col]
y = full_train[target_col]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfclf = RandomForestClassifier()
rfclf.fit(X_train,y_train)
pred = rfclf.predict(X_test)
```

```python
from sklearn.metrics import roc_auc_score
print(roc_auc_score(y_test,pred))
```

```python
from xgboost import XGBClassifier
xgbclf = XGBClassifier()
xgbclf.fit(X_train,y_train)
xgb_pred = xgbclf.predict(X_test)
```

```python
print(roc_auc_score(y_test,xgb_pred))
```

```python
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier()
mlp.fit(X_train,y_train)
mlp_pred = mlp.predict(X_test)
```

```python
print(roc_auc_score(y_test,mlp_pred))
```
