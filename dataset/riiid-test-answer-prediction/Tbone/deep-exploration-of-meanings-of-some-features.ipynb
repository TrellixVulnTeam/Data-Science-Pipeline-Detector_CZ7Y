## Main Data
### task_container_id

``task_container_id`` is used to id which couple qustions/lectures are showed at the same time. As you can see below that all three questions happened at the same timestamp.

```python
train_df[40:43]
```

### user_answer
``user_answer`` has only 5 unique values. Since the data description only says ``the user's answer to the question``, my opinion is that this value may indicate some kinds of categories or measurement of the answer. It's not very realistic that there are only 4 answers to all these questions. Now I said it. It is possible that all the questions are multiple choice/multiple response questions.

```python
train_df['user_answer'].unique()
```

Upon checking unique values of ``correct_answer``, every question has only one answer. If the questions are choice questions, they are multiple choice.

```python
questions_df['correct_answer'].unique()
```

## Lectures Data
### tag

There is only one ``tag`` for each lecture.

```python
lectures_df['tag'].unique()
```

### part

``part`` column has 7 unique ids which are compatible with the 7 test parts mentioned in this link [section of the TOEIC test](https://www.iibc-global.org/english/toeic/test/lr/about/format.html). It further comfirems that the questions are multiple choice questions, since it mentions ``Mark your answers on your answer sheet``.

```python
lectures_df['part'].unique()
```

## Question Data
### bundle_id

``bundle_id`` is used to id which couple qustions are served together, as shown below. 

```python
questions_df.loc[questions_df.duplicated('bundle_id', keep=False),:][:3]
```

Now we choose ``question_id`` 1400 and find one of the users who take question 1400. Observing it's neighbour records, we can see questions belong the same bundle show at the same time. They have the same ``task_container_id``, but its value is different with ``bundle_id``.

```python
train_df.iloc[1023702:1023702+5,:]
```

Let's check one more user just to be safe. We can reach the same conclution from below data. Moreover, eventhough these two users received the same bundle, the values of `` task_container_id`` are different.

```python
train_df.iloc[94636590:94636590+5,:]
```
