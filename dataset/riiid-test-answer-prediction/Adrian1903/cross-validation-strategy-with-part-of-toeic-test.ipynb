I am considering a cross validation ‚úÖ on this project, with ‚è≤ TimeSeriesSplit(). 

We have a time series which is timestamp. We can consider a second variable, which can be considered as a time series, parts of TOEIC test. We can isolate every block. 
At least for the written (5, 6, 7) and oral parts (1, 2, 3, 4).

Below my approach :

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("default")
from sklearn.model_selection import TimeSeriesSplit
```

```python
%%time

path = "../input/cleaning-riiid-dataset-add-part-of-toeic-test/reduced_riiid_train.pkl.gzip"
train = pd.read_pickle(path)
train = train.astype({'row_id': 'int64',
                      'timestamp': 'int64',
                      'user_id': 'int32',
                      'content_id': 'int16',
                      'content_type_id': 'int8',
                      'task_container_id': 'int16',
                      'user_answer': 'int8',
                      'answered_correctly': 'int8',
                      'prior_question_elapsed_time': 'float32',
                      'prior_question_had_explanation': 'boolean'})
train.info()
```

```python
train = train.sort_values(by=['part', 'timestamp'])

col = ['part', 'timestamp', 'row_id', 'user_id', 'content_id', 'content_type_id',
       'task_container_id', 'user_answer', 'answered_correctly',
       'prior_question_elapsed_time', 'prior_question_had_explanation',
       'question_id']
train[col].set_index('row_id')
```

I want to split like this below ‚¨áÔ∏è‚¨áÔ∏è

```python
# Initialisation du TSCV
n_iter = 5
tscv = TimeSeriesSplit(n_splits=n_iter)

# Echantillonnage
train_lst = []
val_lst = []

for train_index, val_index in tscv.split(train):
    train_lst.append(train_index[-1])
    val_lst.append(val_index[-1] - val_index[0])
    
    # X_train, X_test = X.iloc[train_index], X.iloc[val_index]
    # y_train, y_test = y.iloc[train_index], y.iloc[val_index]

train_tpl = tuple(train_lst)
val_tpl = tuple(val_lst)

# Tracage de l'√©chantillon
ind = np.arange(n_iter) + 1
width = 0.35

btrain = plt.barh(ind, train_tpl, width)
bval = plt.barh(ind, val_tpl, width, left=train_tpl)

plt.yticks(ind)
plt.gca().invert_yaxis()
plt.legend((btrain, bval), ("train", "val"))
plt.title('Sample Cross Validation')
plt.ylabel('Interaction CV')
plt.xlabel('Index')
plt.savefig('./img_cross_validation.png', transparent=True)
plt.show()
```

In my opinion üí≠ it is necessary to treat the parts individually in CV, because :
- there are several levels of difficulty,
- the sizes of the parts are heterogeneous,
- these are exercises to listen to and read.
