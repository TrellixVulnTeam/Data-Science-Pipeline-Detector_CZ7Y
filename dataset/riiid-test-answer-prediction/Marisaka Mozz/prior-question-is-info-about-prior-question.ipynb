prior_question_elapsed_time and prior_question_had_explanation are information about "prior" question.
Therefore, these features can be used by shifting them by -1.

```python
import pandas as pd
```

```python
DTYPES_TRAIN = {
    'row_id': 'int64',
    'timestamp': 'int64',
    'user_id': 'int32',
    'content_id': 'int16',
    'content_type_id': 'int8',
    'task_container_id': 'int16',
    'user_answer': 'int8',
    'answered_correctly':'int8',
    'prior_question_elapsed_time': 'float32',
    'prior_question_had_explanation': 'boolean'
}
usecols=['user_id', 'answered_correctly', 'prior_question_elapsed_time', 'prior_question_had_explanation']
```

```python
train = pd.read_csv('/kaggle/input/riiid-test-answer-prediction/train.csv', dtype=DTYPES_TRAIN, usecols=usecols)
train = train[train.answered_correctly != -1]
```

```python
train['elapsed_shift-1'] = train[['user_id', 'prior_question_elapsed_time']].groupby('user_id').shift(-1)
train['explained_shift-1'] = train[['user_id', 'prior_question_had_explanation']].groupby('user_id').shift(-1)
```

```python
train.dropna(inplace=True)
train['explained_shift-1'] = train['explained_shift-1'].astype('boolean')
```

```python
train[['answered_correctly', 'prior_question_elapsed_time', 'elapsed_shift-1']].groupby('answered_correctly').mean()
```

The average value before shifting are almost the same for answered_correctly 0 and 1, but they differ significantly when shifting.

```python
train[['answered_correctly', 'prior_question_had_explanation', 'explained_shift-1']].groupby('answered_correctly').mean()
```

prior_question_had_explanation doesn't seem to change when shifted.
