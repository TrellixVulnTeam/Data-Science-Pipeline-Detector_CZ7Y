```python
import numpy as np
import pandas as pd
```

```python
ROOT = "/kaggle/input/riiid-test-answer-prediction"
```

```python
questions = pd.read_csv(f"{ROOT}/questions.csv")
lectures = pd.read_csv(f"{ROOT}/lectures.csv")
```

Question Tags One hot encode

```python
tags_lst = []
for tags in questions[~questions["tags"].isnull()]["tags"]:
    tags_lst += tags.split()
    
tags_lst = list(map(int, set(tags_lst)))
tags_lst.sort()

len(tags_lst), max(tags_lst)
```

```python
lst = []
for question_id, tags in questions[["question_id", "tags"]].values:
    ohe = np.zeros(188)
    if str(tags) != "nan":
        for tag in tags.split():
            ohe += np.eye(188)[int(tag)]
    
    lst.append([question_id]+ohe.tolist())
tags_df = pd.DataFrame(lst, columns=["question_id"]+[f"q_tag_{i}" for i in range(188)]).astype(int)
```

```python
tags_df.head()
```

```python
tags_df.to_csv("question_tags_ohe.csv", index=None)
```

Lectures Tag One hot encode

```python
type_of_dic = {v: i for i, v in enumerate(lectures["type_of"].unique())}
type_of_dic
```

```python
lst = []
for lecture_id, tag, type_of in lectures[["lecture_id", "tag", "type_of"]].values:
    row = [lecture_id, type_of_dic[type_of]] + np.eye(188)[tag].tolist()
    lst.append(row)
l_tag_df = pd.DataFrame(lst, columns=["lecture_id", "type_of"] + [f"l_tag_{i}" for i in range(188)]).astype(int)
```

```python
l_tag_df.head()
```

```python
l_tag_df.to_csv("lecture_tags_ohe.csv", index=None)
```
