```python

y = train_data.pop("target")
y = encoder.fit_transform(y)

X = train_data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=1, stratify=y)
```

```python
model = RandomForestClassifier(n_estimators=300, max_depth=15, n_jobs=-1)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_train)
print(y_pred)
```

```python
test_data = test_data.iloc[:, 1:]
test_preds = model.predict_proba(test_data)
submission_data = pd.read_csv("/kaggle/input/tabular-playground-series-jun-2021/sample_submission.csv")
submission_data.iloc[:, 1:] = test_preds
submission_data.to_csv("submission.csv", index=False)
```
