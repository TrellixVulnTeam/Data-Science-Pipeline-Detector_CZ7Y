```python
train = pd.read_csv('../input/tabular-playground-series-jun-2021/train.csv')
test = pd.read_csv('../input/tabular-playground-series-jun-2021/test.csv')
train['target'].hist(figsize=(120,80))
```

```python
X = train[train.columns[1:-1]].values
test_X = test[test.columns[1:]].values
y = train.target.values
```

```python
X.shape, test_X.shape, y.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(bootstrap=True, verbose=1, n_jobs=-1)
rf.fit(X, y)
```

```python
rf.classes_
```

```python
y_pred = rf.predict_proba(test_X)
```

```python
sub = pd.read_csv('../input/tabular-playground-series-jun-2021/sample_submission.csv')
```

```python
sub[sub.columns[1:]] = y_pred
```

```python
sub.to_csv('submission.csv', index=False)
```
