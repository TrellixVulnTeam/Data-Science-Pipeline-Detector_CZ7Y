```python
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
train_data = TabularDataset('/kaggle/input/tabular-playground-series-jun-2021/train.csv').drop('id', axis=1)
train_data.head()
```

```python
test_data = TabularDataset('/kaggle/input/tabular-playground-series-jun-2021/test.csv').drop('id', axis=1)
test_data.head()
```

```python
metric = 'log_loss'
```

```python
predictor = TabularPredictor(label='target', eval_metric=metric).fit(train_data= train_data, time_limit=14400, presets='best_quality', verbosity=2)
predictor.leaderboard(train_data, silent=True)
```

```python
results = predictor.fit_summary()
```

```python
pred_probs = predictor.predict_proba(test_data, model='CatBoost_BAG_L1')
pred_probs.head(5)
```
