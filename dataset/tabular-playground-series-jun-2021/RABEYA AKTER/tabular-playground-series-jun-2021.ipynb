```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore', category=DeprecationWarning)
warnings.filterwarnings('ignore', category=FutureWarning)
warnings.filterwarnings('ignore', category=RuntimeWarning)
warnings.filterwarnings('ignore', category=UserWarning)

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost.sklearn import XGBClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import ExtraTreesClassifier
from catboost import CatBoostClassifier
import lightgbm as lgb
from sklearn import metrics

!pip install catboost
```

```python
train=pd.read_csv("../input/tabular-playground-series-jun-2021/train.csv")
train.head()
```

```python
test=pd.read_csv("../input/tabular-playground-series-jun-2021/test.csv")
test.head()
```

```python
sample_submission=pd.read_csv("../input/tabular-playground-series-jun-2021/sample_submission.csv")
sample_submission.head()
```

```python
print('train:             ',len(train))
print('test:              ',len(test))
print('sample submission: ',len(sample_submission))
```

```python
print('train-null:             ',train.isnull().sum().sum())
print('test-null:              ',test.isnull().sum().sum())
print('sample submission-null: ',sample_submission.isnull().sum().sum())
```

```python
train.drop(['id'],axis=1,inplace=True)
x=train.drop(['target'],axis=1,inplace=False).values
y=train['target'].values
x_train,x_val,y_train,y_val=train_test_split(x,y,test_size=0.25)
```

```python
def classifier():
  #logistic regression
  clf1= LogisticRegression()
  print("                       Logistic Regression")
  clf1.fit(x_train,y_train)
  test_pred=clf1.predict(x_val)
  print(metrics.classification_report(y_val, test_pred))
  print('\n')

  #Random Forest Classifier
  clf2= RandomForestClassifier()
  print("                       Random Forest Classifier")
  clf2.fit(x_train,y_train)
  test_pred=clf2.predict(x_val)
  print(metrics.classification_report(y_val, test_pred))
  print('\n')

  #XGBoost Classifier
  clf3= XGBClassifier(eval_metric='mlogloss')
  print("                       XGBoost Classifier")
  clf3.fit(x_train,y_train)
  test_pred=clf3.predict(x_val)
  print(metrics.classification_report(y_val, test_pred))
  print('\n')

  #AdaBoost Regression
  clf4= AdaBoostClassifier()
  print("                       AdaBoost Classifier")
  clf4.fit(x_train,y_train)
  test_pred=clf4.predict(x_val)
  print(metrics.classification_report(y_val, test_pred))
  print('\n')

  #Extra Trees Classifier
  clf5= ExtraTreesClassifier()
  print("                       Extra Trees Classifier")
  clf5.fit(x_train,y_train)
  test_pred=clf5.predict(x_val)
  print(metrics.classification_report(y_val, test_pred))
  print('\n')

  #Decision Tree Classifier
  clf6= DecisionTreeClassifier()
  print("                       Decision Tree Classifier")
  clf6.fit(x_train,y_train)
  test_pred=clf6.predict(x_val)
  print(metrics.classification_report(y_val, test_pred))
  print('\n')

  #LGBM Classifier
  clf7= lgb.LGBMClassifier()
  print("                       LGBM Classifier")
  clf7.fit(x_train,y_train)
  test_pred=clf7.predict(x_val)
  print(metrics.classification_report(y_val, test_pred))
  print('\n')

  #Cat Boost Classifier
  clf8= CatBoostClassifier(logging_level='Silent')
  print("                       Cat Boost Classifier")
  clf8.fit(x_train,y_train)
  test_pred=clf8.predict(x_val)
  print(metrics.classification_report(y_val, test_pred))
  print('\n')

classifier()  
```

```python
clf = ExtraTreesClassifier()
clf.fit(x, y)
```

```python
test=test.drop(['id'],axis=1).values
prediction=clf.predict_proba(test)
submission=pd.DataFrame(prediction,
                  columns=['Class_1','Class_2','Class_3','Class_4','Class_5','Class_6','Class_7','Class_8','Class_9'])
submission['id']=sample_submission['id']
id= submission.pop('id')
submission.insert(0,'id',id)
```

```python
submission.to_csv("submission.csv",index=False)
submission.head()
```
