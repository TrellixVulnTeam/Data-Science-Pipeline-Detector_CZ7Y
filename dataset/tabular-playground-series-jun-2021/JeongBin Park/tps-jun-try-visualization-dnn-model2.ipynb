```python
# Using library
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from keras import layers, models

import matplotlib.font_manager as fm
font = fm.FontProperties(fname='../input/staatfont/Staatliches-Regular.ttf')
```

```python
train = pd.read_csv('../input/tabular-playground-series-jun-2021/train.csv')
test = pd.read_csv('../input/tabular-playground-series-jun-2021/test.csv')

X = train.drop(['target','id'],axis=1)
y = train['target']
test = test.drop(['id'],axis=1)
```

### Dimensionality reduction. Because, It has so many columns. So I reduce the columns using PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=27, svd_solver='randomized', whiten =True)
X = pca.fit_transform(X)

X = np.array(X)
y = np.array(pd.get_dummies(y))
test = np.array(test)
```

Build Model - DNN
</h1>
</div>

```python
model = models.Sequential()
model.add(layers.Dense(9, input_shape=X.shape, activation='relu'))
model.add(layers.Dense(9, activation='softmax'))
model.compile(loss = 'categorical_crossentropy', optimizer='adam')
model.summary()
```

### I just build so simple model. This time, I focus on visualization !

```python
history = model.fit(X, y, epochs=30, batch_size=256, validation_split=0.3,verbose=1)
```

```python
weight = model.get_weights()
layer1 = pd.DataFrame([list(range(1,28,1)),np.ones((27))]).T
layer2 = pd.DataFrame([list(range(1,28,3)),np.ones((9))]).T
layer2[1] = layer2[1]*2
layer3 = layer2.copy()
layer3[1] = 3
point = pd.concat([layer1,layer2,layer3]).reset_index(drop=True)
point.columns = ['y', 'x']
```

```python
print("weight value shape : ",model.get_weights()[0].shape, model.get_weights()[2].shape)
print("bias value shape : ", model.get_weights()[1].shape, model.get_weights()[3].shape)
```

```python
fig = plt.figure(figsize=(20,13))
plt.scatter( x= point['x'],y=point['y'], s=400,color='black')

weight_i = 0
weight_j =0
for i in range(1,28):
    weight_j =0
    for j in range(1,28,3):
        tmp = pd.DataFrame({'x':[1,2], 'y': [i,j]})
        we = weight[0][weight_i][weight_j]
        if we >= 0:
            color = 'green'
        else:
            color = 'red'
        sns.lineplot(x=tmp['x'],y=tmp['y'],ci=None,lw=np.abs(we)**2*2,color=color)
        weight_j = weight_j+1
    weight_i = weight_i+1
    
weight_i = 0
weight_j =0
for i in range(1,28,3):
    weight_j =0
    for j in range(1,28,3):
        tmp = pd.DataFrame({'x':[2,3], 'y': [i,j]})
        we = weight[2][weight_i][weight_j]
        if we >= 0:
            color = 'green'
        else:
            color = 'red'
        sns.lineplot(x=tmp['x'],y=tmp['y'],lw=np.abs(we)**2*2,ci=None,color=color)
        weight_j = weight_j+1
    weight_i = weight_i+1
        
plt.axis('off')
plt.text(s='Input_layer', x=1,y=28,font = font, fontsize=18,va='center',ha='center')
plt.text(s='hidden_layer', x=2,y=28,font = font,fontsize=18,va='center',ha='center')
plt.text(s='out_layer', x=3,y=28,font = font,fontsize=18,va='center',ha='center')
plt.text(s='Consider only weight value', x=0.9,y=30,va='center',ha='left',font = font,fontsize=50)

plt.show()
```

```python
fig = plt.figure(figsize=(20,13))
ax = fig.add_subplot(111)
plt.scatter( x= point['x'],y=point['y'], s=400,color='black')

weight_i = 0
weight_j =0
for i in range(1,28):
    weight_j =0
    for j in range(1,28,3):
        tmp = pd.DataFrame({'x':[1,2], 'y': [i,j]})
        we = weight[0][weight_i][weight_j]
        bias = weight[1][weight_j]
        value = we + bias
        if value >= 0:
            color = 'green'
        else:
            color = 'red'
        sns.lineplot(x=tmp['x'],y=tmp['y'],ci=None,lw=np.abs(we)**2*2,color=color)
        weight_j = weight_j+1
    weight_i = weight_i+1
    
weight_i = 0
weight_j =0
for i in range(1,28,3):
    weight_j =0
    for j in range(1,28,3):
        tmp = pd.DataFrame({'x':[2,3], 'y': [i,j]})
        we = weight[2][weight_i][weight_j]
        bias = weight[3][weight_j]
        value = we + bias
        if value >= 0:
            color = 'green'
        else:
            color = 'red'
        sns.lineplot(x=tmp['x'],y=tmp['y'],lw=np.abs(we)**2*2,ci=None,color=color)
        weight_j = weight_j+1
    weight_i = weight_i+1
        
        
plt.axis('off')
plt.text(s='Input_layer', x=1,y=28,font = font, fontsize=18,va='center',ha='center')
plt.text(s='hidden_layer', x=2,y=28,font = font,fontsize=18,va='center',ha='center')
plt.text(s='out_layer', x=3,y=28,font = font,fontsize=18,va='center',ha='center')
plt.text(s='Consider weight and bias value', x=0.9,y=30,va='center',ha='left',font = font,fontsize=50)
plt.text(s='Consider Relu', x=2,y=27,font = font,fontsize=18,va='center',ha='center')


ax.axvspan(1.8, 2.2, fc="gray", alpha=0.2)

plt.show()
```

### Then how about consider activation function like ReLu? 
### Activation function can't visualization totally so, I will trying visualization using Only one data

```python
fig = plt.figure(figsize=(20,13))
ax = fig.add_subplot(111)
plt.scatter( x= point['x'],y=point['y'], s=400,color='black')

weight_i = 0
weight_j =0
for i in range(1,28):
    weight_j =0
    for j in range(1,28,3):
        tmp = pd.DataFrame({'x':[1,2], 'y': [i,j]})
        we = weight[0][weight_i][weight_j]
        bias = weight[1][weight_j]
        value = we + bias
        if value >= 0:
            color = 'green'
        else:
            color = 'red'
        sns.lineplot(x=tmp['x'],y=tmp['y'],ci=None,lw=np.abs(we)**2*2,color=color)
        weight_j = weight_j+1
    weight_i = weight_i+1
    
weight_i = 0
weight_j =0
for i in range(1,28,3):
    weight_j =0
    for j in range(1,28,3):
        tmp = pd.DataFrame({'x':[2,3], 'y': [i,j]})
        we = weight[2][weight_i][weight_j]
        bias = weight[3][weight_j]
        value = we + bias
        if value >= 0:
            color = 'green'
        else:
            color = 'red'
        sns.lineplot(x=tmp['x'],y=tmp['y'],lw=np.abs(we)**2*2,ci=None,color=color)
        weight_j = weight_j+1
    weight_i = weight_i+1
        
for i in range(1,28):
    value = round(X[0][i-1],3)
    plt.text(s =value, x=0.9, y=i,font=font, fontsize=18,va='center',ha='center')

tmp_arr = []
for i in range(0,9):
    value = 0
    for j in range(0,27):
        value = value + X[0][j] * weight[0][j][i] + weight[1][i]
    tmp_arr.append(value)
    value =round(value,3)
    plt.text(s = value, x=1.9,y=i*3+1, font=font, fontsize=18,va='center',ha='center')
    
for i in range(0,9):
    if tmp_arr[i] > 0:
        value = tmp_arr[i]
    else:
        tmp_arr[i] = 0
        value = 0
    value =round(value,3)
    plt.text(s = value, x=2.1,y=i*3+1, font=font, fontsize=18,va='center',ha='center')

tmp_arr2 = []
for i in range(0,9):
    value = 0
    for j in range(0,9):
        value = value + tmp_arr[j] * weight[2][j][i] + weight[3][i]
    value =round(value,3)
    tmp_arr2.append(value)
    plt.text(s = value, x=2.9,y=i*3+1, font=font, fontsize=18,va='center',ha='center')

ax.axvspan(1.8, 2.2, fc="gray", alpha=0.2)
ax.axvspan(2.8, 3.2, fc="gray", alpha=0.2)

def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum()
soft = softmax(tmp_arr2)

for i in range(0,9):
    value =round(soft[i],3)
    plt.text(s = value, x=3.1,y=i*3+1, font=font, fontsize=18,va='center',ha='center')
    
plt.axis('off')
plt.text(s='Input_layer', x=1,y=28,font = font, fontsize=18,va='center',ha='center')
plt.text(s='hidden_layer', x=2,y=28,font = font,fontsize=18,va='center',ha='center')
plt.text(s='out_layer', x=3,y=28,font = font,fontsize=18,va='center',ha='center')
plt.text(s='DNN Model', x=0.9,y=30,va='center',ha='left',font = font,fontsize=50)
plt.text(s='Consider Relu', x=2,y=27,font = font,fontsize=18,va='center',ha='center')
plt.text(s='Consider Softmax', x=3,y=27,font = font,fontsize=18,va='center',ha='center')

plt.show()
```
