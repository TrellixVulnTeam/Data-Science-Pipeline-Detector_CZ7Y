```python
# Using library
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from keras import layers, models
```

```python
train = pd.read_csv('../input/tabular-playground-series-jun-2021/train.csv')
test = pd.read_csv('../input/tabular-playground-series-jun-2021/test.csv')

X = train.drop(['target','id'],axis=1)
y = train['target']
test = test.drop(['id'],axis=1)
```

### Dimensionality reduction. Because, It has so many columns. So I reduce the columns using PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=27, svd_solver='randomized', whiten =True)
X = pca.fit_transform(X)

X = np.array(X)
y = np.array(pd.get_dummies(y))
test = np.array(test)
```

Build Model - DNN
</h1>
</div>

```python
model = models.Sequential()
model.add(layers.Dense(9, input_shape=X.shape, activation='relu'))
model.add(layers.Dense(9, activation='softmax'))
model.compile(loss = 'categorical_crossentropy', optimizer='adam')
model.summary()
```

### I just build so simple model. This time, I focus on visualization !

```python
history = model.fit(X, y, epochs=30, batch_size=256, validation_split=0.3,verbose=1)
```

```python
weight = model.weights
layer1 = pd.DataFrame([list(range(1,28,1)),np.ones((27))]).T
layer2 = pd.DataFrame([list(range(1,28,3)),np.ones((9))]).T
layer2[1] = layer2[1]*2
layer3 = layer2.copy()
layer3[1] = 3
point = pd.concat([layer1,layer2,layer3]).reset_index(drop=True)
point.columns = ['y', 'x']
```

```python
fig = plt.figure(figsize=(20,13))
plt.scatter( x= point['x'],y=point['y'], s=400,color='black')

weight_i = 0
weight_j =0
for i in range(1,28):
    weight_j =0
    for j in range(1,28,3):
        tmp = pd.DataFrame({'x':[1,2], 'y': [i,j]})
        we = weight[0][weight_i][weight_j]
        if we >= 0:
            color = 'green'
        else:
            color = 'red'
        sns.lineplot(x=tmp['x'],y=tmp['y'],ci=None,lw=np.abs(we)**2*2,color=color)
        weight_j = weight_j+1
    weight_i = weight_i+1
    
weight_i = 0
weight_j =0
for i in range(1,28,3):
    weight_j =0
    for j in range(1,28,3):
        tmp = pd.DataFrame({'x':[2,3], 'y': [i,j]})
        we = weight[2][weight_i][weight_j]
        if we >= 0:
            color = 'green'
        else:
            color = 'red'
        sns.lineplot(x=tmp['x'],y=tmp['y'],lw=np.abs(we)**2*2,ci=None,color=color)
        weight_j = weight_j+1
    weight_i = weight_i+1
        
plt.axis('off')
plt.text(s='Input_layer', x=1,y=28,fontsize=18,va='center',ha='center')
plt.text(s='hidden_layer', x=2,y=28,fontsize=18,va='center',ha='center')
plt.text(s='out_layer', x=3,y=28,fontsize=18,va='center',ha='center')

plt.show()
```

### I using only model's weight. ( Ignore bias ) 
### If weight is postive number, Green line, If weight is negative number, Red line
### And, Using abs(weight)**2, Make their's width ! - Because, If set only abs(weight), I can't compare their's weight
