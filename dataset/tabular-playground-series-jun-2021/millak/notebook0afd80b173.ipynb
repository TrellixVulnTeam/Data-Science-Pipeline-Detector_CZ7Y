```python
# Using library
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from keras import layers, models
```

```python
train = pd.read_csv('../input/tabular-playground-series-jun-2021/train.csv')
test = pd.read_csv('../input/tabular-playground-series-jun-2021/test.csv')

X_train = train.drop(['target','id'],axis=1)
y_train = train['target']
X_test = test.drop(['id'],axis=1)
```

```python
len(X_train.columns)
```

```python
model = models.Sequential()
model.add(layers.Dense(9, input_shape=X_train.shape, activation='relu'))
model.add(layers.Dense(9, activation='softmax'))
model.compile(loss = 'categorical_crossentropy', optimizer='adam')
model.summary()
```

```python
X = np.array(X_train)
y = np.array(pd.get_dummies(y_train))
```

```python
X.shape,y.shape
```

```python
history = model.fit(X, y, epochs=30, batch_size=256, validation_split=0.3,verbose=1)
```

```python
predictions = model.predict(X_test)
```

```python
predictions
```

```python
ans_df = pd.DataFrame(predictions,
 columns = ['Class_1','Class_2','Class_3','Class_4','Class_5','Class_6','Class_7','Class_8','Class_9'])
```

```python
ans_df
```

```python
ans_df['id']=test['id'].values
```

```python
output = ans_df[['id','Class_1','Class_2','Class_3','Class_4','Class_5','Class_6','Class_7','Class_8','Class_9']]
output.to_csv('submission.csv', index=False)
```
