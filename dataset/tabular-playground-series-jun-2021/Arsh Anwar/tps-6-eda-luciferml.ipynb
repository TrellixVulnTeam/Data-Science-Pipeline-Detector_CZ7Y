<h3 style="text-align:center;font-weight: bold">Correlation Plot</h3></div>



```python
corr = train.corr()
mask = np.triu(np.ones_like(corr, dtype = bool))
plt.figure(figsize = (15, 15))
plt.title('Correlation matrix Train Set')
sns.heatmap(corr, mask = mask, cmap = 'rocket', linewidths = .5)
plt.show()
```

```python
corr = test.corr()
mask = np.triu(np.ones_like(corr, dtype = bool))
plt.figure(figsize = (15, 15))
plt.title('Corelation matrix Test Set')
sns.heatmap(corr, mask = mask, cmap = 'rocket', linewidths = .5)
plt.show()
```

### From above plot we can see there is very high correlation b/w some features and very less in some features

<h3 style="text-align:center;font-weight: bold">Feature Distribution</h3></div>



```python
fig = plt.figure(figsize = (20, 70))
for i in tqdm_notebook(range(len(train.columns)), desc = 'Your Charts are being ready'):
    fig.add_subplot(np.ceil(len(train.columns)/5), 5, i+1)
    train.iloc[:, i].hist(bins = 20)
    plt.title('Train feature_{}'.format(i))
plt.show()
```

```python
fig = plt.figure(figsize = (20, 70))
for i in tqdm_notebook(range(len(test.columns)), desc = 'Your Charts are being ready'):
    fig.add_subplot(np.ceil(len(test.columns)/5), 5, i+1)
    train.iloc[:, i].hist(bins = 20)
    plt.title('Test feature_{}'.format(i))
plt.show()
```

### From above plots we can see columns are skewed
