```python
test = pd.read_csv('/kaggle/input/tabular-playground-series-jun-2021/test.csv')
train = pd.read_csv('/kaggle/input/tabular-playground-series-jun-2021/train.csv')
sample_submission = pd.read_csv('/kaggle/input/tabular-playground-series-jun-2021/sample_submission.csv')
```

```python
train_X = train.drop(['target', 'id'], axis = 1)
train_X.head()
```

```python
train_y = train['target']
train_y.head()
```

```python
test_X = test.drop('id', axis = 1)
test_X
```

```python
classes = list(set(train_y))
classes.sort()
print(classes)
```

```python
import lightgbm as lgb

lgb_cls = lgb.LGBMClassifier()
lgb_cls.fit(train_X, train_y)
```

```python
pred = lgb_cls.predict_proba(test_X)
```

```python
pred[:5]
```

```python
submission = pd.DataFrame(pred, columns = classes, index = sample_submission['id'])
```

```python
submission.head()
```

```python
submission.to_csv('submission.csv')
```
