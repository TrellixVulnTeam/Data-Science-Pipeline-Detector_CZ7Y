```python
X = data.drop(['target', 'id'], axis=1)
y = data['target']
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
y = to_categorical(y)
```

```python
from sklearn.decomposition import PCA
# Make an instance of the Model
pca = PCA(30)
```

```python
pca.fit(X)
```

```python
transformed_x = pca.transform(X)
# transformed_y = pca.transform(y)
```

```python
transformed_x.shape
```

```python
import tensorflow as tf
```

```python
model = tf.keras.models.Sequential(
    [tf.keras.layers.Dense(300, input_shape=[30], activation='relu'),
    tf.keras.layers.Dense(150, activation='relu'),
    tf.keras.layers.Dense(150, activation='relu'),
    tf.keras.layers.Dense(150, activation='relu'),
    tf.keras.layers.Dense(9, activation='softmax')]
)
    
                      
                      
```

```python
model.compile(optimizer='sgd', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
X.shape
```

```python
len(y)
```

```python
model.fit(transformed_x,y,epochs=50,batch_size=35)
```

```python
test = pd.read_csv('/kaggle/input/tabular-playground-series-jun-2021/test.csv')
```

```python
pre_test = test.drop(['id'],axis=1)
```

```python
pre_test = pca.transform(pre_test)
```

```python
prediction = model.predict(pre_test)
```

```python
output = pd.DataFrame(prediction,columns=['Class_1', 'Class_2', 'Class_3', 'Class_4', 'Class_5', 'Class_6',
       'Class_7', 'Class_8', 'Class_9'])
```

```python
output['id'] = test['id']
```

```python
output.to_csv('My_submission_1.csv',index=False)
```
