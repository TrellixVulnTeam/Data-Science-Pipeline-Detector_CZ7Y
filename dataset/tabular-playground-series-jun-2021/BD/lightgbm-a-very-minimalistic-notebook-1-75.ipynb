The idea behind this notebook is to make a super simple base for beginners, and demonstrating the power of Lightgbm.

```python
# Importing libraries
import numpy as np
import pandas as pd
import lightgbm as lgb
```

```python
# Loading the data
train=pd.read_csv('../input/tabular-playground-series-jun-2021/train.csv')
test=pd.read_csv('../input/tabular-playground-series-jun-2021/test.csv')
```

```python
# Preprocessing!!
features=train.columns[1:-1]
X=train[features]
y=train['target']

classes=list(set(train['target']))
classes.sort()
```

```python
# fitting the model and predicting the probabilities
LGB=lgb.LGBMClassifier()
LGB.fit(X,y)
pred=LGB.predict_proba(test[features])
```

```python
# submission
df=pd.DataFrame(pred,columns=classes,index=test['id'])
df.to_csv('submission.csv')
```
