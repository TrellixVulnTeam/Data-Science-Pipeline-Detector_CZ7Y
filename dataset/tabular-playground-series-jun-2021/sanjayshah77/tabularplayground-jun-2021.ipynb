```python
df=pd.read_csv('../input/tabular-playground-series-jun-2021/train.csv')
```

```python
df.head()
```

```python
print('number of classes: ',len(df.target.unique()))
```

```python
embarked_dummies=pd.get_dummies(df.target,prefix='class')
```

```python
embarked_dummies.head()
```

```python
df = pd.concat([df, embarked_dummies], axis=1)
```

```python
df.head()
```

```python
df=df.drop(['id','target'],axis=1)
```

```python
df.head()
```

```python
val_df = df.sample(frac=0.2, random_state=1337)
train_df = df.drop(val_df.index)

print(
    "Using %d samples for training and %d for validation"
    % (len(train_df), len(val_df))
)
```

```python
label_list=embarked_dummies.columns.tolist()
```

```python
df_labels=train_df[label_list]
```

```python
df_labels.head()
```

```python
df_features=train_df.copy()
```

```python
df_features=df_features.drop(label_list,axis=1)
```

```python
df_features.head()
```

```python
# Use the tutorial below to work from here.  
#https://keras.io/examples/structured_data/structured_data_classification_from_scratch/
```

```python
# Make numpy values easier to read.
np.set_printoptions(precision=3, suppress=True)

import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.layers.experimental import preprocessing
```

```python
features = np.array(df_features)
features
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(128, activation='relu'))
model.add(tf.keras.layers.Dense(64, activation='relu'))
model.add(tf.keras.layers.Dense(32, activation='relu'))
model.add(tf.keras.layers.Dense(9, activation='softmax'))
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.categorical_crossentropy,
              metrics=['accuracy'])
```

```python
df_features_val=val_df.copy()
```

```python
df_features_val=df_features_val.drop(label_list,axis=1)
```

```python
features_val = np.array(df_features_val)
features_val
```

```python
df_labels_val=val_df[label_list]
```

```python
loss, accuracy = model.evaluate(x=features_val,y=df_labels_val)
print("Accuracy", accuracy)
```

```python

from keras import callbacks
earlystopping = callbacks.EarlyStopping(monitor ="val_loss", 
                                        mode ="min", patience = 5, 
                                        restore_best_weights = True)
  
history = model.fit(x=features, y=df_labels, 
                    epochs = 25, validation_data =(features_val, df_labels_val), 
                    callbacks =[earlystopping])
```
