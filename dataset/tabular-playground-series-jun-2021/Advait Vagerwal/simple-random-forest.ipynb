# creating the model

```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model
X = train.drop(['target'], axis = 'columns', inplace = False).values
y = train['target'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 10)
test_submit = test.drop(['id'], axis =1).values
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
clf = RandomForestClassifier(n_estimators = 100)
clf.fit(X, y)
#Now, finding the accuracy score
y_pred = clf.predict(X_test)
print(accuracy_score(y_test,y_pred)*100)
```

# Creating the submission file

```python
sample = pd.read_csv('../input/tabular-playground-series-jun-2021/sample_submission.csv')
print(sample.head(2))
```

```python
y_result= clf.predict_proba(test_submit)
submission = pd.DataFrame(y_result,columns=['Class_1','Class_2','Class_3','Class_4', 'Class_5', 'Class_6', 'Class_7', 'Class_8', 'Class_9'])
submission['id'] = sample['id']
print(submission.head())
```
