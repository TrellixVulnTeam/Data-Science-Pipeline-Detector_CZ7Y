```python
test = pd.read_csv('../input/tabular-playground-series-jun-2021/test.csv')
train = pd.read_csv('../input/tabular-playground-series-jun-2021/train.csv')
train.head(2)
```

```python
test.head(2)
```

```python
train = train.drop(['id'], axis = 1)
test = test.drop(['id'], axis =1)
```

```python
train.isna().sum().sum()
```

```python
train.info()
```

```python
from sklearn.model_selection import train_test_split
X = train.drop(['target'], axis = 'columns', inplace = False)
y = train['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 10)
```

```python
from xgboost import XGBClassifier
clf = XGBClassifier(random_state=1, max_depth=8, n_estimators = 20000, learning_rate = 0.001, verbosity = 0)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred)*100)
```

```python
pred = clf.predict_proba(X_test)
```

```python
sample = pd.read_csv('../input/tabular-playground-series-jun-2021/sample_submission.csv')
print(sample.head(2))
```

```python
y_result= clf.predict_proba(test)
submission = pd.DataFrame(y_result,columns=['Class_1','Class_2','Class_3','Class_4', 'Class_5', 'Class_6', 'Class_7', 'Class_8', 'Class_9'])
submission['id'] = sample['id']
submission.to_csv('submission.csv',index=False)
submission
```
