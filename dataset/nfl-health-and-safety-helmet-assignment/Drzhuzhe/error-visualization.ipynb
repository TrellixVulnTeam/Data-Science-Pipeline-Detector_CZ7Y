```python
import pandas as pd
from helmet_assignment.video import video_with_predictions
from IPython.display import Video, display
```

```python
SOURCE_DIR = '../input/nfl-health-and-safety-helmet-assignment/'
PREDICT_DIR = '../input/helper-code-helmet-mapping-deepsort/'

# submission
train_label = pd.read_csv(f'{SOURCE_DIR}/train_labels.csv')
submission = pd.read_csv(f'{PREDICT_DIR}/submission.csv')
```

```python
submission.head()
```

```python
train_label.head()
```

```python
submission['video'] = submission.video_frame.str.split("_").str[:3].str.join("_") + ".mp4"
submission['frame'] = submission.video_frame.str.split("_").str[-1].astype('int')
```

```python
submission.head()
```

```python
#debug_label = "57783_003374_Endzone_1"
#debug_videos = "57783_003374_Endzone.mp4"
#train_label.loc[train_label.video_frame == debug_label]
debug_videos = submission['video'].unique()
```

```python
debug_labels = train_label.query('video in @debug_videos')
scorer = NFLAssignmentScorer(debug_labels)
scorer.score(submission)
```

```python
# Create video showing predictions for one of the videos.
frac = 0.60 # scaling factor for display

for debug_video in debug_videos:
    video_out = video_with_predictions(
        f'{SOURCE_DIR}/train/{debug_video}',
        scorer.sub_labels)


    display(Video(data=video_out,
                  embed=True,
                  height=int(720*frac),
                  width=int(1280*frac))
           )
```
