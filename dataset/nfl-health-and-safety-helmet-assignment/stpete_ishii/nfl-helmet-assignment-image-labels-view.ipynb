```python
def draw_box(num0):
    
    impath=impaths[num0]
    image=cv2.imread(impath)
    file=impath.split('/')[-1]
    box=image_labels[image_labels['image']==file].iloc[:,1:6].reset_index(drop=True)
    #print(box)
    
    for i in range(len(box)):
        label=box.iloc[i,0]
        x=box.iloc[i,1]
        w=box.iloc[i,2]
        y=box.iloc[i,3]
        h=box.iloc[i,4]
        
        if label==labels[0]:
            cv2.rectangle(image,(x,y),(x+w,y+h),(255,255,0),2)  # yellow 'Helmet'
        elif label==labels[1]:
            cv2.rectangle(image,(x,y),(x+w,y+h),(0,255,255),2)  # light blue 'Helmet-Blurred            
        elif label==labels[2]:
            cv2.rectangle(image,(x,y),(x+w,y+h),(255,0,0),2)  # red 'Helmet-Difficult'
        elif label==labels[3]:
            cv2.rectangle(image,(x,y),(x+w,y+h),(0,255,0),2)  # yellow green 'Helmet-Sideline'               
        elif label==labels[4]:
            cv2.rectangle(image,(x,y),(x+w,y+h),(0,0,255),2)  # blue 'Helmet-Partial'             
            
        #cv2.imwrite('/kaggle/working/source0/'+file+'.png',image)        
        #plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
        #plt.show() 
        
    return image
```

```python
images1=[]
for i in tqdm(range(len(impaths)//10)):
    images1+=[draw_box(i)]
```

```python
from matplotlib import animation, rc
rc('animation', html='jshtml')
```

```python
def create_animation(ims):
    
    fig=plt.figure(figsize=(12,6))
    plt.axis('off')
    im=plt.imshow(cv2.cvtColor(ims[0],cv2.COLOR_BGR2RGB))
    
    def animate_func(i):
        im.set_array(cv2.cvtColor(ims[i],cv2.COLOR_BGR2RGB))
        return [im]

    return animation.FuncAnimation(fig, animate_func, frames=len(ims), interval=1000//2)
```

```python
create_animation(images1)
```
