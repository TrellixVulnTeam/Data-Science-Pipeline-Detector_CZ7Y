```python
# import dependencies
import os
import subprocess
import shutil
```

```python
def video_process(source_dir_path, dst_dir_path, types, img_ext, maxSize=1024):
    if not os.path.exists(dst_dir_path):
        os.makedirs(dst_dir_path)

    # check that video formats equal types 
    for file_name in os.listdir(source_dir_path):
        for type in types:
            if type not in file_name:
                continue

        name, ext = os.path.splitext(file_name)
    
        video_file_path = os.path.join(source_dir_path, file_name)

        # skip large files
        # if os.path.getsize(video_file_path) > maxSize * 1000:
        #	continue
        
        # get frames
        cmd = 'ffmpeg -i \"{}\" -qscale:v 2 \"{}/{}_%d.{}\"'.format(video_file_path, dst_dir_path, name, img_ext)

        print(cmd)
        subprocess.call(cmd, shell=True)
        print('\n')
```

```python
# create destination folder for images
img_ext = 'png'
dst_dir_path = f'/kaggle/working/{img_ext}_from_mp4/'
os.makedirs(dst_dir_path, exist_ok=True)

# source folder with videos
source_dir_path = '../input/nfl-health-and-safety-helmet-assignment/test'

# allowed video types 
types = ['.mp4']

# convert
video_process(source_dir_path, dst_dir_path, types, img_ext)
```

## Convert back from PNG to MP4

```python
import cv2
import os

video_name = 'video.mp4'
video_dir_path = f'/kaggle/working/mp4_from_{img_ext}/'
os.makedirs(video_dir_path, exist_ok=True)

frame_rate = 38 

# get only frames for 57995_000109_Endzone_ video
images = [img for img in os.listdir(dst_dir_path) if img.startswith("57995_000109_Endzone_")]

# need to sort images to create video from this sorted list
sorted_images = [''] * len(images)
for im in images:
    filename, _ = os.path.splitext(im.split('_')[-1])
    sorted_images[int(filename) - 1] = im
```

```python
# get image size
frame = cv2.imread(os.path.join(dst_dir_path, images[0]))
height, width, layers = frame.shape

# create video
video = cv2.VideoWriter(os.path.join(video_dir_path, video_name), cv2.VideoWriter_fourcc(*'DIVX'), frame_rate, (width,height))

for image in sorted_images:
    video.write(cv2.imread(os.path.join(dst_dir_path, image)))

video.release()
```
