```python
import cv2
```

```python
df = pd.read_csv("../input/nfl-health-and-safety-helmet-assignment/train_labels.csv")
```

```python
len(df.video.unique())
```

```python
frames_dir = "frames"
os.makedirs(frames_dir,exist_ok=True)
```

```python
for video in df.video.unique():
    source = f"../input/nfl-health-and-safety-helmet-assignment/train/{video}"
    basename = os.path.basename(source)
    basename = os.path.splitext(basename)[0]
        
    cap = cv2.VideoCapture(source)
    i = 1
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        # resize image
        #frame = cv2.resize(frame, (WIDHT, HEIGHT), interpolation=cv2.INTER_AREA)
        frame_path = os.path.join(frames_dir, f"{basename}_{i}.jpg")
        cv2.imwrite(frame_path, frame)
        print(f"{i}-->{basename}", end="\r")
        i += 1
```
