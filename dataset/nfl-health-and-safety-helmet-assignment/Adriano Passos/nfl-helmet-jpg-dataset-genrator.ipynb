# NFL-Helmet JPG Dataset Genrator

This notebook converts all video files into a series of JPG images. The name of the resulting images are in the format `game_play_view_frame`, for example: `./train/57686_002546_Endzone_048.jpg`.

The dataset generated using this code can be found on: https://www.kaggle.com/coldfir3/nflhelmet-jpg-dataset

```python
import os
import shutil
from zipfile import ZipFile
from pathlib import Path
from tqdm.notebook import tqdm
```

```python
def zip_folder(folder, rm_original = True):
    # iterate over all the files in directory
    for folderName, subfolders, filenames in os.walk(folder):
        # create a ZipFile object
        with ZipFile(folderName.split('/')[-1] + '.zip', 'a') as zipObj:
            for filename in filenames:
                # create complete filepath of file in directory
                filePath = os.path.join(folderName, filename)
                # add file to zip
                zipObj.write(filePath, os.path.basename(filePath))
                # delete the file to open space
                if rm_original:
                    os.remove(filePath)
    if rm_original:
        shutil.rmtree(folder)
```

```python
try:
    shutil.rmtree('train')
except:
    pass
os.mkdir('train')
z = ZipFile("train.zip", "w")
z.close()
```

```python
path = Path('../input/nfl-health-and-safety-helmet-assignment/train')
for filename in tqdm(path.iterdir(), total = len(list(path.iterdir()))):
    os.system(f"ffmpeg -i ../input/nfl-health-and-safety-helmet-assignment/train/{filename.stem}.mp4 train/{filename.stem}_%03d.jpg")

zip_folder('train')
```
