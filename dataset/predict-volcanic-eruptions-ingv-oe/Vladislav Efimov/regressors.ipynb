```python
import numpy as np 
import pandas as pd 
import os
import time
```

```python
train_df = pd.read_csv('../input/vulcanic-preprocessing/train_sample.csv')
test_df = pd.read_csv('../input/vulcanic-preprocessing/test.csv')
test_df.drop(test_df.columns[0], axis=1, inplace=True)
```

```python
train_x = train_df.to_numpy()
test_x = test_df.to_numpy()
train_y = pd.read_csv('../input/vulcanic-preprocessing/targets.csv').to_numpy()
train_y = train_y.ravel()
```

```python

train_x.shape, test_x.shape, train_y.shape
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.feature_selection import SelectKBest,f_regression
from sklearn.pipeline import Pipeline

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.ensemble import GradientBoostingRegressor
import xgboost as xgb
import lightgbm as lgb
```

```python
rf = RandomForestRegressor()
rf.n_jobs = -1
start = time.time()
rf.fit(train_x, train_y)
rf_score = cross_val_score(rf,train_x,train_y,scoring= 'neg_mean_squared_error', cv=3)
rf_train_rmse_score = np.sqrt(-rf_score)

train_r2_score = rf.score(train_x, train_y) 

end = time.time()
elapsed = (end - start) // 60, (end - start) % 60  
print(f'train RMSE: {rf_train_rmse_score.mean()}, train R2: {train_r2_score}, elapsed time: {elapsed[0]} mins, {elapsed[1]} secs')
    
    
```

```python
tree = DecisionTreeRegressor()
tree.n_jobs = -1
start = time.time()
tree.fit(train_x, train_y)
tree_score = cross_val_score(tree,train_x,train_y,scoring= 'neg_mean_squared_error', cv=3)
tree_train_rmse_score = np.sqrt(-tree_score)

train_r2_score = tree.score(train_x, train_y) 

end = time.time()
elapsed = (end - start) // 60, (end - start) % 60  
print(f'train RMSE: {tree_train_rmse_score.mean()}, train R2: {train_r2_score}, elapsed time: {elapsed[0]} mins, {elapsed[1]} secs')
```

```python
xgb_model = xgb.XGBRegressor()
start = time.time()
xgb_model.fit(train_x, train_y)
xgb_score = cross_val_score(xgb_model,train_x,train_y,scoring= 'neg_mean_squared_error', cv=3)
xgb_train_rmse_score = np.sqrt(-xgb_score)

train_r2_score = xgb_model.score(train_x, train_y) 

end = time.time()
elapsed = (end - start) // 60, (end - start) % 60  
print(f'train RMSE: {xgb_train_rmse_score.mean()}, train R2: {train_r2_score}, elapsed time: {elapsed[0]} mins, {elapsed[1]} secs')
```

```python
lgb_model = lgb.LGBMRegressor()
start = time.time()
lgb_model.fit(train_x, train_y)
lgb_score = cross_val_score(lgb_model,train_x,train_y,scoring= 'neg_mean_squared_error', cv=3)
lgb_train_rmse_score = np.sqrt(-lgb_score)

train_r2_score = lgb_model.score(train_x, train_y) 

end = time.time()
elapsed = (end - start) // 60, (end - start) % 60  
print(f'train RMSE: {lgb_train_rmse_score.mean()}, train R2: {train_r2_score}, elapsed time: {elapsed[0]} mins, {elapsed[1]} secs')
```
