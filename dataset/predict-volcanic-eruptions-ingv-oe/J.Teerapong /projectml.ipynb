```python
import numpy as np
import pandas as pd
import os
from keras.models import Sequential
from keras.layers import GRU,RepeatVector,Dense,LSTM,TimeDistributed,Dropout
from sklearn.preprocessing import StandardScaler;
import matplotlib.pyplot as plt
from tensorflow.keras import optimizers
```

```python
data_dir = '../input/predict-volcanic-eruptions-ingv-oe' 
```

```python

def show_sensors(df1,df2):
    f, axes = plt.subplots(10, 2)
    f.set_size_inches((16, 10)) 
    f.tight_layout()
    plt.subplots_adjust(bottom=-0.4)
    for i in range(1,11):
        axes[i-1][0].plot(df1[i-1].values,c = 'r')
        axes[i-1][0].set_title('Sensor_'+str(i))
        axes[i-1][0].set_xlabel('time')
    for i in range(1,11):
        axes[i-1][1].plot(df2[i-1].values,c = 'b')
        axes[i-1][1].set_title('Sensor_'+str(i))
        axes[i-1][1].set_xlabel('time')
```

```python
Xmodel = Sequential()
Xmodel.add(LSTM(100,input_shape=(501,10)))
Xmodel.add(Dense(1))
optimizers.SGD(learning_rate=0.000005)
Xmodel.compile(optimizer='sgd',loss='mean_squared_error')
Xmodel.summary()
```

re Y

```python
y_rescaler = StandardScaler();
y = pd.read_csv('../input/dataml/Y_sp500.csv');
y = y_rescaler.fit_transform(y)
y.shape
```

re X

```python
x_rescaler = StandardScaler();
x = pd.read_csv('../input/dataml/X_sp500.csv');
x = x.values.reshape(4431,501,10)
x1 = x
for i in x:
    x_rescaler.fit(i);
x = x_rescaler.transform(x.reshape(4431*501,10))
x = x.reshape(4431,501,10)
```

read csv

```python
i = 0
show_sensors(pd.DataFrame(x[i]),pd.DataFrame(x1[i]))
```

```python
x_train = x[:4000]
x_test = x[4000:]
y_train = y[:4000]  
y_test = y[4000:]
```

```python
y_train.shape
```

```python
print(x.shape,y.shape)
hist = Xmodel.fit(x_train, y_train,validation_split = 0.1, epochs=50,batch_size=1)
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train_loss, val_loss'], loc='upper left')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(Xmodel.predict(x_train),y_train))
print(mean_squared_error(Xmodel.predict(x_test),y_test))
```
