# Training Data

The main file that serves as an index of all training data is train.csv.
This file has two columns, *segment_id* and *time_to_eruption*.
Each segment_id points to another training data csv file.
The time_to_eruption is what we want to predict.

```python
train_index = pd.read_csv('/kaggle/input/predict-volcanic-eruptions-ingv-oe/train.csv')
print(f"Columns in train.csv: {list(train_index.columns)}")
print(f"Number of segments in training: {len(train_index)}")
print("*" * 60)
print("Training Index Data:")
print(train_index.head())
```
