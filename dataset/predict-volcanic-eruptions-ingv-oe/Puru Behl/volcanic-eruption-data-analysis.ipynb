# Importing The Packages 

```python
import pandas as pd
import numpy as np
from IPython.display import clear_output
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

## Importing the data

```python
data=pd.read_csv('../input/predict-volcanic-eruptions-ingv-oe/train.csv')
```

### Having a look at the data

```python
data.head()
```

### Code to get some standard data instead of all the data

```python
l=len(data.segment_id)
X=[]
count=0
for i in data.segment_id:
    clear_output(wait=True)
    df=pd.read_csv('../input/predict-volcanic-eruptions-ingv-oe/train/'+str(i)+'.csv')
    df=df.fillna(0)
    X.append(df.mean().values)
    count+=1
    print('The current progress :',"{:.2f}".format(count*100/l),' %')
```

```python
np.save('Volcanid_main_data.py',X)
```

Also added the progress calculator

This data has the most of the features like mean , std , min ,max etc of all the sensors

```python
y=data['time_to_eruption'].values
```

```python
X=np.array(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4)
```

```python
model=LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
X_train.shape
```

```python
print('The error in the data is :',mean_absolute_error(model.predict(X_test),y_test))
```

```python
from sklearn.cross_decomposition import PLSRegression
```

```python
model2=PLSRegression(n_components=1,tol= 0.0005009294958980265)
```

```python
model2.fit(X_train,y_train)
```

```python
print('The error of this model is :',mean_absolute_error(model2.predict(X_test),y_test))
```

Error of this model is lower than the error of the simple linear regression model as we can see
