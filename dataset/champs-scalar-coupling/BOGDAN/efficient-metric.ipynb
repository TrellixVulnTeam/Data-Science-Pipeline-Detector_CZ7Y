```python
import numpy as np
import pandas as pd
```

```python
%%time
train = pd.read_csv("../input/train.csv")

y_train = train['scalar_coupling_constant']
groups = train['type']
```

```python
def group_mean_log_mae(y_true, y_pred, groups, floor=1e-9):
    maes = (y_true-y_pred).abs().groupby(groups).mean()
    return np.log(maes.map(lambda x: max(x, floor))).mean()
```

```python
# zeros prediction
group_mean_log_mae(y_train, 0, groups)
```

```python
# best possible prediction
group_mean_log_mae(y_train, y_train, groups)
```

### Check Speed

```python
%timeit group_mean_log_mae(y_train, y_train, groups)
```
