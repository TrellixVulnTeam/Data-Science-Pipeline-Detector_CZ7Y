```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(train_data['cuisine'])

train_data['encoded']=le.transform(train_data['cuisine'])
```

```python
train_data.head()
```

```python
from sklearn.model_selection import train_test_split
X=train_data.drop(['id','ingredients','cuisine','encoded'],axis=1)
y=train_data['encoded']
```

```python
X_train,X_val,y_train,y_val=train_test_split(X,y,random_state=42)
```

```python
X_train[:100][:]
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train[0:4000][:],y_train[0:4000][:])
```

```python
lr.fit(X_train[5000:9000][:],y_train[5000:9000][:])
lr.score(X_val,y_val)
```

```python
test_data.head()
```

```python
test_data['encoded']=lr.predict(test_data.drop(['id','ingredients'],axis=1))
test_data['cuisine']=le.inverse_transform(test_data['encoded'])
```

```python
Submission=test_data[['id','cuisine']]
Submission.set_index('id',inplace=True)
```
