Basicamente, temos um dataset completo. 
Como a proposta é de fazer uma classificação binária e temos que entender, na análise, o que define cada uma das classificações, vamos dividir o dataset em dois para fins de comparação


```python
df0 = df[df['target'] == 0].drop(['id', 'target'], axis=1).reset_index(drop=True)
df1 = df[df['target'] == 1].drop(['id', 'target'], axis=1).reset_index(drop=True)
```

```python
dfcor1 =  df1.corr(method ='pearson')
dfcor1.head()
```

Vamos para a correlação com a classificação

```python
dfcor[['target']].sort_values(by='target')
```

```python
ax = sns.heatmap(dfcor*100)
```

As correlações são muito baixas para ter algo significativo para a visualização interessate.
Mas vamos tentar apenas uma parte desses dados:

Pegando a diferença de cada divisão de acordo com o df principal: 

em outras palavras, os maiores valores mostram as correlações mais diferentes (logo, mais significativas de diferenças entre os grupos)

```python
scor = abs((dfcor - dfcor0) - (dfcor - dfcor1))
scor
```

```python
(scor > 0.01).sum().sum()
```

O limiar é bem baixo, difícil saber se é realmente significativo

```python
plt.figure(figsize=(20, 20))
sns.heatmap(scor, cbar=False)
```

```python
plt.figure(figsize=(20, 50))
sns.heatmap(scor[:100], cbar=False)
```

```python
plt.figure(figsize=(20, 50))
sns.heatmap(scor[101:200], cbar=False)
```

```python
plt.figure(figsize=(20, 50))
sns.heatmap(scor[201:], cbar=False)
```

Dá para ver que são poucos os pontos brancos

Normalizando...

```python
dfcor_n = pd.DataFrame(
                  MinMaxScaler().fit_transform(scor), 
                  index = scor.index, 
                  columns=scor.columns)

dfcor_n.head()
```

Vamos ver tirando a coluna wheezy-copper-turtle-magic

```python
dfr = df.drop(['id', 'wheezy-copper-turtle-magic', 'target'], axis=1)
```

```python
dfr0 = df[df['target'] == 0].drop(['id', 'target', 'wheezy-copper-turtle-magic'], axis=1).reset_index(drop=True)
dfr1 = df[df['target'] == 1].drop(['id', 'target', 'wheezy-copper-turtle-magic'], axis=1).reset_index(drop=True)
```

```python
corfr= dfr.corr(method ='pearson')
corfr
```

```python
corfr0= dfr0.corr(method ='pearson')
corfr0
```

```python
corfr1= dfr1.corr(method ='pearson')
corfr1
```

```python
scor2 = abs((corfr - corfr0) - (corfr - corfr1) )
scor2
```

```python
dfcor_nt = pd.DataFrame(
                  MinMaxScaler().fit_transform(scor2), 
                  index = scor2.index, 
                  columns=scor2.columns)

dfcor_nt.head()
```

```python
plt.figure(figsize=(25, 25))
sns.heatmap(dfcor_nt, cbar=False)
```
