```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error
from sklearn.preprocessing import MinMaxScaler
```

```python
submission=pd.read_csv("/kaggle/input/tabular-playground-series-aug-2021/sample_submission.csv")
submission.head()
```

```python
actual=submission['loss'].values
```

```python
train=pd.read_csv("/kaggle/input/tabular-playground-series-aug-2021/train.csv")
test=pd.read_csv("/kaggle/input/tabular-playground-series-aug-2021/test.csv")
train.shape,test.shape
```

```python
train.head()
```

```python
train.describe()
```

```python
train.isna().any()
test.isna().any()
```

```python
columns=train.columns[1:-1]
columns
```

```python
xtrain=train[columns].values
ytrain=train["loss"].values.reshape(-1,1)
testx=test[columns].values
```

```python
model=LinearRegression()
```

```python
model.fit(xtrain,ytrain)
out=model.predict(testx)
print(mean_absolute_error(actual,out))
out
```

```python
submission['loss']=out
```
