# Adjusting data in preparation for building models:

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
target=train_data["loss"]
features=train_data.drop(["loss"],axis=1)
```

```python
#MinMaxScaler for Data
scaler = MinMaxScaler(copy=True, feature_range=(0, 1))
featur_train= scaler.fit_transform(features)
```

```python
featur_train
```

# Building models with machine learning:

# Linear Regression algorithm

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression(fit_intercept=True, normalize=True,copy_X=True,n_jobs=-1)
model.fit(featur_train, target)
```

```python
pred=model.predict(test)
```

```python
result = pd.DataFrame({'id': submition.id,
                       'loss': pred, 
                       })

result.to_csv('submission.csv', index = False)
result
```
