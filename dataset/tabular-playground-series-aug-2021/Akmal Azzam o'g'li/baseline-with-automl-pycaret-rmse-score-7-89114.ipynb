```python
def pycaret_model(train, target, test, n_select, fold, opt, include):
  print('Setup Your Data....')
  setup(data=train,
        target=target,
        ignore_features = ['id'],
        #transform_target = True,
        normalize = True,
        #numeric_imputation = 'mean',
        silent= True)
  
  print('Comparing Models....')
  best = compare_models(sort=opt, n_select=n_select, fold = fold, include = include)

  print('Blending Models....')
  #blended = blend_models(estimator_list= best, fold=fold, optimize=opt)
  pred_holdout = predict_model(best)

  print('Finallizing Models....')
  final_model = finalize_model(best)

  print('Done...!!!')

  pred_esb = predict_model(final_model, test)
  re = pred_esb['Label']

  return re
```

```python
%%time
sample_sub['loss'] = pycaret_model(train, 'loss', test, 1, 2, 'RMSE', ['catboost'])
```
