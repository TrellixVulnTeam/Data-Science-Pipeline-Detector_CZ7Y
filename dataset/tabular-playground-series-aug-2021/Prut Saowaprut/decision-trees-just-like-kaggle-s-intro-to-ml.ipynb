```python
import pandas as pd
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
```

## Load the data

```python
train = pd.read_csv('../input/tabular-playground-series-aug-2021/train.csv')
test = pd.read_csv('../input/tabular-playground-series-aug-2021/test.csv')
```

Let's have a look at our data.

```python
train.head()
```

```python
X = train.iloc[:,1:-1]
y = train.loss
```

Split the data into train and validation sets.

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, train_size=0.8, random_state=1)
```

Making a function which predicts and then calculates the mae.

```python
def get_mae(nodes):
    model = DecisionTreeRegressor(max_leaf_nodes=nodes, random_state=1)
    model.fit(X_train, y_train)
    preds = model.predict(X_val)
    mae = mean_absolute_error(preds, y_val)
    return mae
```

Let's try different numbers of max leaf nodes for our tree.

```python
n_nodes_l = [5, 10, 50, 100, 200]
maes=[]
for n in n_nodes_l:
    maes.append(get_mae(n))

best_n_index = maes.index(min(maes))
best_n = n_nodes_l[best_n_index]
```

```python
best_n
```

The best number was 10, let's now use all of our data to train our final model.

```python
final_model = DecisionTreeRegressor(max_leaf_nodes=10, random_state=1)
final_model.fit(X,y)
```

```python
test_preds = final_model.predict(test.iloc[:,1:])
```

That's it! Thank you very much.

```python
output = pd.DataFrame({'id': test.id,
                       'loss': test_preds})
output.to_csv('submission.csv', index=False)
```
