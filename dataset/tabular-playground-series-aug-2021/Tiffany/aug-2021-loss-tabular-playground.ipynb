```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

path="../input/tabular-playground-series-aug-2021/train.csv"
```

```python
csv = pd.read_csv(path)
```

```python
X = csv.drop("loss", axis=1)
y = csv.loss
```

```python
from sklearn.model_selection import train_test_split
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)
```

```python
from sklearn.tree import DecisionTreeRegressor
iowa_model = DecisionTreeRegressor(random_state=1)
iowa_model.fit(train_X, train_y)
```

```python
from sklearn.metrics import mean_squared_error
val_predictions = iowa_model.predict(val_X)
val_mse = mean_squared_error(val_predictions, val_y)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
for max_leaf_nodes in [10, 50, 100, 200, 500, 1000]:
    rf_model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=1)
    rf_model.fit(train_X, train_y)
    rf_val_predictions = rf_model.predict(val_X)
    rf_val_mse = mean_squared_error(rf_val_predictions, val_y)

    print(rf_val_mse)
```

```python
import pandas as pd

testX = pd.read_csv("../input/tabular-playground-series-aug-2021/test.csv")

rf_model = DecisionTreeRegressor(max_leaf_nodes=10, random_state=1)
rf_model.fit(train_X, train_y)
test_preds = rf_model.predict(testX)
```
