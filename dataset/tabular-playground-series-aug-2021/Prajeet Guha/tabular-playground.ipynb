```python
sample_submission = pd.read_csv('/kaggle/input/tabular-playground-series-aug-2021/sample_submission.csv')
sample_submission.head()
```

```python
train_data = pd.read_csv('/kaggle/input/tabular-playground-series-aug-2021/train.csv')
train_data.head()
```

```python
train_data.info()
```

```python
train_data.describe()
```

```python
X = train_data.drop(['loss','id'], axis = 1 )
y = train_data['loss']
scalar = preprocessing.StandardScaler().fit(X)
X = scalar.transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
lr = LinearRegression()
lr.fit(X_train,y_train)
lr.score(X_test,y_test)
```

```python
result_feature = pd.read_csv('/kaggle/input/tabular-playground-series-aug-2021/test.csv')
predict_label = scalar.transform(result_feature.drop('id',axis = 1))
```

```python
prediction = lr.predict(predict_label)
sample_submission['loss'] = prediction
```
