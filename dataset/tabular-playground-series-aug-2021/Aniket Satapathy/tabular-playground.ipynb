***MACHINE LEARNING MODEL*** 

```python
x_train = train_data.drop('loss', axis=1)
y_train = train_data['loss']
x_test = test_data
y_test = pd.read_csv('../input/tabular-playground-series-aug-2021/sample_submission.csv')
```

```python
id_loc = np.array(y_test['id'])
```

```python
y_test = y_test.drop('id', axis=1)
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

Dimentionality reduction

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
lda = LDA(n_components = 10)
x_train = lda.fit_transform(x_train, y_train)
x_test = lda.transform(x_test)
```

XGBoost Model

```python
from xgboost import XGBRegressor
regressor = XGBRegressor()
regressor.fit(x_train, y_train)
```

```python
y_pred = regressor.predict(x_test)
```

```python
len(y_pred)
type(y_pred)
```

```python
output = pd.DataFrame({'id':id_loc, 'loss':y_pred})
```
