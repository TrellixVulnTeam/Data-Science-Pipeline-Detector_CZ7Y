### Our idea is to "Snap to Curve" based on a "Guide Point". The best "Guide Point" is determined based on the evaluation equation "RMSE". The value "np.mean (yy)" is appropriate here, but we do not know what that number is. So first we consider "np.mean (y)" and then we modify its value with a number "R".

### Obviously, the values of the coefficients "coeff1" and "coeff2" must be greater than one. Because we want to do "Snap to Curve". That is, we want to replace the percentage of a value (point on the curve) with the same percentage of "Guide_Point". But coefficients smaller than one are usually used for averaging.

## Guide Point

```python
guide = np.mean(y)
# np.sum(y) / len(y)
display(guide)
```

## Coordinate

> #### [ main ] : Result Received.

> #### [ guide ] : Guide Point.

> #### [ coeff1 ] : The coefficient for any value is smaller than the Guide_Point.

> #### [ coeff2 ] : The coefficient for any value is greater than the Guide_Point.

> #### [ R ] : Value modification for Guide_Point.

```python
def coordinate(main, guide, coeff1, coeff2, R):   
    
    sub  = main.copy() 
    subv = sub.values    
    suba = subv[:, 1]
    
    coor  = main.copy()    
    coorv = coor.values
    guide = guide - R
    
    for i in range (len(main)):
        
        if (suba[i] <= guide):
            per = (suba[i] * coeff1) + (guide * (1.0 - coeff1))
            coorv[i, 1] = per
            
        if (suba[i] > guide):
            per = (suba[i] * coeff2) + (guide * (1.0 - coeff2)) 
            coorv[i, 1] = per        
        
    coor.iloc[:, 1] = coorv[:, 1] 
    
    ###############################   
    X = suba
    Y = coor.iloc[:, 1]
    g = guide   
    
    plt.style.use('seaborn-whitegrid') 
    plt.figure(figsize=(9, 9), facecolor='lightgray')
    plt.title(f'\nC O O R D I N A T E\n')   
            
    plt.scatter(X, X, s=2.0, label='Main(X=Y)')
    plt.scatter(X, Y, s=2.0, label='Coordinated-1')
    plt.scatter(g, g, s=100, label='Guide') 
    
    plt.legend(fontsize=12, loc=2)
    #plt.savefig('Coordinate_1.png')
    plt.show()     
    ###############################   
    coor.iloc[:, 1] = coor.iloc[:, 1].astype(float)
    hist_data = [suba, coor.iloc[:, 1]] 
    group_labels = ['Main', 'Coordinated-1']
    
    fig = ff.create_distplot(hist_data, group_labels, bin_size=.2, show_hist=False, show_rug=False)
    fig.show()   
    ###############################       
    print()
    print(':::::::::::::::: Main Values ::::::::::::::::')
    print(f'Min:{suba.min()}   Max:{suba.max()}\n')
    print(':::::::::::: Coordinated Values :::::::::::::')
    print(f'Min:{coor.iloc[: ,1].min()}   Max:{coor.iloc[: ,1].max()}\n')     
    ###############################    
    
    return coor   
```

```python
sub = coordinate(sub_received, guide, 1.16, 1.10 , 0.50)
```
