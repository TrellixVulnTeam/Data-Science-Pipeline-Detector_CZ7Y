```python
def r(p):
    return pd.read_csv(p)
def wc(f1, f2):
    f1 = f1 - 84000
    f2 = f2 - 84000
    cal = f2 / (f1 + f2)
    print("weight", cal)
    return cal
def g(s1, s2):
    f1 = int(s1.split("/")[-1].split(".")[1].split(" ")[0])
    f2 = int(s2.split("/")[-1].split(".")[1].split(" ")[0])
    w = wc(f1, f2)
    file1 = r(s1)
    file2 = r(s2)
    file1.loss = file1.loss * w + file2.loss * (1 - w)
    file1.to_csv(f'file1_7.{f1}_file2_7.{f2}_blend.csv',index=False)
    return file1
```
