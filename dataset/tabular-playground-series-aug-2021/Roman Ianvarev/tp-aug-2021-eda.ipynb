No missing values, no need to impute nothing.

```python
c_mat = train.corr()
print('matrix_rank(cmat) - {}, det(c_mat) = {}'.format(np.linalg.matrix_rank(c_mat), np.linalg.det(c_mat)))
c_mat.shape
```

```python
for i in range(0, len(c_mat)):
    c_mat.iloc[i,i] = 0 #eliminating main diagonal filled with ones
print(c_mat.max().max(), c_mat.min().min())
```

We making a very brief conclusion that nothing is collinear in this features, and nothing corellates. Let's get a closer look to distributions.
