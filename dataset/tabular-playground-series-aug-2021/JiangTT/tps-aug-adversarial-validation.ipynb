```python
from supervised import AutoML
from sklearn.model_selection import train_test_split
```

```python
train_df = pd.read_csv('../input/tabular-playground-series-aug-2021/train.csv')
test_df = pd.read_csv('../input/tabular-playground-series-aug-2021/test.csv')
```

```python
train_df['target'] = 0
test_df['target'] = 1
train = pd.concat([train_df.drop(columns=['id', 'loss']).sample(test_df.shape[0], random_state=42), test_df.drop(columns=['id'])], axis=0)
train, test = train_test_split(train, test_size=0.2, random_state=42)
```

```python
automl = AutoML(
    mode='Explain',
    ml_task='binary_classification',
    algorithms=['LightGBM', 'Xgboost', 'CatBoost'],
    eval_metric='accuracy',
    random_state=42,
)
automl.fit(train.drop(columns=['target']), train['target'])
```

```python
automl.score(test.drop(columns=['target']), test['target'])
```

```python
automl.report()
```
