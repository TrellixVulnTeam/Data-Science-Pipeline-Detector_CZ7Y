```python
import scipy
import os
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import ggplot
import plotly
import seaborn as sns
```

```python
train_data = pd.read_csv('../input/tabular-playground-series-aug-2021/train.csv', index_col= False)
train_data.head()
```

```python
train_data.corr()
```

```python
plt.figure(figsize=(20, 15))
sns.heatmap(train_data.corr(), linecolor ='black', cmap='RdYlBu')
plt.show()
```

```python
plt.figure(figsize=(20, 20))
sns.heatmap(data= train_data.groupby('loss').mean().sort_index(), vmin=-0.5, vmax=0.5, cmap=sns.diverging_palette(240, 10, as_cmap=True), linecolor='black', linewidth=1)
plt.show()
```

```python
plt.figure(figsize=(20, 20))
sns.heatmap(data= train_data.groupby('loss').std().sort_index(), vmin=-0.5, vmax=0.5, cmap=sns.diverging_palette(240, 10, as_cmap=True), linecolor='black', linewidth=1)
plt.show()
```
