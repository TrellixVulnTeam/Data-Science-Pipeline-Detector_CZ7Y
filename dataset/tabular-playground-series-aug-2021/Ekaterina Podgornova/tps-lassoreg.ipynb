### Data preparation 

```python
X = train.drop(['id', 'loss'], axis = 1)
Y = train['loss']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
   X, Y, test_size=0.3, random_state=42
)
```

```python
sc=StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
#df_test = StandardScaler().fit_transform(test.drop('id', axis = 1))
```

### Linear regression

```python
model_lr = LinearRegression().fit(X_train, y_train)
model_lr.score(X_test, y_test)
```

```python
y_pred_lr = model_lr.predict(X_test)
print('mean_absolute_error: ', mean_absolute_error(y_test, y_pred_lr))
print('mean_squared_error: ', np.sqrt(mean_squared_error(y_test, y_pred_lr)))
```

#### Lasso regression

```python
lasso_model = Lasso(alpha=1.0).fit(X_train, y_train)
lasso_model.score(X_test, y_test)
```

```python
y_pred_lasso = lasso_model.predict(X_test)
print('mean_absolute_error: ', mean_absolute_error(y_test, y_pred_lasso))
print('mean_squared_error: ', np.sqrt(mean_squared_error(y_test, y_pred_lasso)))
```

```python
finpreds2 = lasso_model.predict(test.drop('id', axis=1))
```

```python
submission = pd.read_csv('/kaggle/input/tabular-playground-series-aug-2021/sample_submission.csv')
submission["loss"]= finpreds2
submission.to_csv("submission2.csv", index=False)
```
