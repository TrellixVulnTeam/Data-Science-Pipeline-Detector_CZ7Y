```python
train = pd.read_csv('../input/tabular-playground-series-aug-2021/train.csv')
test = pd.read_csv('../input/tabular-playground-series-aug-2021/test.csv')
sample_submission = pd.read_csv('../input/tabular-playground-series-aug-2021/sample_submission.csv')
```

```python
sample_submission.head()
train.head()
test.head()
```

```python
#train,testをひとまとめにする	特徴量エンジニアリングの際にtestデータを考慮した標準化を行えるなどの利点があるが、実際には予測対象のデータを得ていない状態でモデルを作成するので、testデータを使った標準化モデルの作成はしないほうがいいとの意見もある。
data = pd.concat([train, test], sort=False)
data.head()
print(len(train), len(test), len(data))
data.isnull().sum()
```

```python
#訓練開始
train = data[:len(train)]
test = data[len(train):]

y_train = train['loss']
X_train = train.drop('loss', axis=1)
X_test = test.drop('loss', axis=1)

X_train.head()
y_train.head()
```

```python
#まずはロジスティック回帰で訓練してみる
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(penalty='l2', solver='sag', random_state=0)	#random_state=0でseedを固定してある
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
y_pred[:20]
```
