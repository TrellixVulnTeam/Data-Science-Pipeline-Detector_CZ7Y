```python
df_train = pd.read_csv("/kaggle/input/tabular-playground-series-aug-2021/train.csv")
df_test = pd.read_csv("/kaggle/input/tabular-playground-series-aug-2021/test.csv")
sub = pd.read_csv("/kaggle/input/tabular-playground-series-aug-2021/sample_submission.csv")
```

```python
df_train = df_train[df_train["loss"] < 10]
```

```python
ff = [c for c in df_test.columns if c not in ['id']]
X = df_train[ff]
y = df_train['loss']
X_test = df_test[ff]
```

```python
SPLITS=10
```

```python
def get_folds(X,Y):
    kf = KFold(SPLITS, shuffle=True)
    folds = np.zeros(X.shape[0], dtype='int16')
    count = 0 
    for fold_n, (train_index, valid_index) in enumerate(kf.split(X)):
        count = count + 1
        folds[valid_index] = count
    return folds


FOLDS = True

if FOLDS:
    folds = get_folds(X,y).reshape(-1,1)
    ids = df_train["id"].to_numpy().reshape(-1,1)
    f = np.append(ids, folds, axis=1)
    ff = pd.DataFrame(data=df_train["id"], columns = ["id"], dtype=np.int16)
    ff["folds"] = folds
    ff.to_csv("tabular_aug_2021_folds_"+str(SPLITS)+".csv")
```
