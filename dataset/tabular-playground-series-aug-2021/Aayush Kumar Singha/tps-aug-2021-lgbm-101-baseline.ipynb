# Imports

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from lightgbm import LGBMRegressor
```

# Load dataset

```python
train = pd.read_csv('../input/tabular-playground-series-aug-2021/train.csv')
test = pd.read_csv('../input/tabular-playground-series-aug-2021/test.csv')
print('train shape:',train.shape)
print('test shape:',test.shape)
```

```python
train.head()
```

```python
# Train data
X_train = train.drop(columns = ['loss','id'])
y_train = train['loss'].values

# Test data
X_test=test.drop(columns = ['id'])
print('Train set:', X_train.shape)
print('Test set:', X_test.shape)
```

# Model training

```python
lgbm_params = {'reg_alpha': 0.4988899107600194, 
               'reg_lambda': 0.3216983283790298, 
               'num_leaves': 64, 
               'learning_rate': 0.03033148284558732, 
               'max_depth': 33, 
               'n_estimators': 5581, 
               'min_child_weight': 0.0166605551617851, 
               'subsample': 0.9566304128710696, 
               'colsample_bytree': 0.6207274451420828, 
               'min_child_samples': 77}
```

```python
model = LGBMRegressor(**lgbm_params)
model.fit(X_train, y_train,
          eval_metric="rmse",
          verbose=True)
```

# Prediction

```python
y_pred = model.predict(X_test)
```

# Submission

```python
preds = pd.read_csv("../input/tabular-playground-series-aug-2021/sample_submission.csv")
preds.loss = y_pred
preds.head()
```

```python
preds.to_csv('submission_lgbm_101.csv', index=False)
```
