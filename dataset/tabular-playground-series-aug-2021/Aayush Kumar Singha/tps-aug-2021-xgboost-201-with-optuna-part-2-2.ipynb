# Imports

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import joblib
from xgboost import XGBRegressor
```

# Load datasets

```python
train = pd.read_csv('../input/tabular-playground-series-aug-2021/train.csv')
test = pd.read_csv('../input/tabular-playground-series-aug-2021/test.csv')
print('train shape:',train.shape)
print('test shape:',test.shape)
```

```python
# Train data
X_train = train.drop(columns = ['loss','id'])
y_train = train['loss'].values

# Test data
X_test=test.drop(columns = ['id'])
print('Train set:', X_train.shape)
print('Test set:', X_test.shape)
```

# Load tuned hyperparameters obtained from [XGBoost 201 - with Optuna Part 1/2](https://www.kaggle.com/aayush26/tps-aug-2021-xgboost-201-with-optuna-part-1-2/notebook)

```python
study = joblib.load("../input/tps-aug-2021-xgboost-201-with-optuna-part-1-2/xgb_study.pkl")
```

# Model training

```python
model = XGBRegressor(
    **study.best_params, 
    tree_method='gpu_hist', 
    learning_rate=0.01, 
    booster='gbtree'
)
```

```python
model.fit(X_train, 
          y_train,
          eval_metric="rmse",
          verbose=True)
```

# Prediction

```python
y_pred = model.predict(X_test)
```

# Submission

```python
preds = pd.read_csv("../input/tabular-playground-series-aug-2021/sample_submission.csv")
preds.loss = y_pred
preds.head()
```

```python
preds.to_csv('submission_xgbboost_201.csv', index=False)
```
