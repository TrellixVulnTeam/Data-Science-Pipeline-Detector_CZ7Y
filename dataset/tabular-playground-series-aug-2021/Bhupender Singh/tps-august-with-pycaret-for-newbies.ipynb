# 1. INITIALIZING THE MODELS

```python
setup(data = train, target= "loss",silent = True, session_id = 2021, use_gpu = True)
```

```python
xgboost = create_model("xgboost")
```

```python
lgbm = create_model("lightgbm")
```

```python
catboost = create_model("catboost")
```

# 2. TUNING THE MODELS

```python
xgboost_tuned = tune_model(xgboost, optimize= "RMSE")
```

```python
lgbm_tuned = tune_model(lgbm, optimize = "RMSE")
```

```python
catboost_tuned = tune_model(catboost, optimize = "RMSE")
```

```python
blended_models = blend_models([xgboost_tuned,lgbm_tuned, catboost_tuned], optimize= "RMSE")
```

```python
final_model = finalize_model(blended_models)
```
