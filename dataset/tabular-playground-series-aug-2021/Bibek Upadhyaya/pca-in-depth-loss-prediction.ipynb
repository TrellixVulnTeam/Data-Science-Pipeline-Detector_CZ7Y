```python
#PCA
```

```python
df.head()
```

```python
#
# Scale the dataset; 
#
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
a=df.drop('loss',axis=1).values
b=df['loss'].values
```

```python
sc.fit(a)
```

```python
a_std = sc.transform(a)
```

```python
from sklearn.decomposition import PCA
pca=PCA()
```

```python
a_pca = pca.fit_transform(a_std)
```

```python
# Determine explained variance using explained_variance_ration_ attribute
#
exp_var_pca = pca.explained_variance_ratio_
```

```python
exp_var_pca
```

```python
# Cumulative sum of eigenvalues; This will be used to create step plot
# for visualizing the variance explained by each principal component.
#
cum_sum_eigenvalues = np.cumsum(exp_var_pca)
```

```python
cum_sum_eigenvalues
```

```python
plt.bar(range(0,len(exp_var_pca)), exp_var_pca, alpha=0.5, align='center', label='Individual explained variance')
plt.step(range(0,len(cum_sum_eigenvalues)), cum_sum_eigenvalues, where='mid',label='Cumulative explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal component index')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

##Train-Test Split

```python
x = df_new.drop("loss",axis=1)
y = df_new[["loss"]]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=49)
print("x_train :",x_train.shape)
print("x_test :",x_test.shape)
print("y_train :",y_train.shape)
print("y_test :",y_test.shape)
```

```python
#Linear Regression
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
preds = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mean_squared_error(y_test,preds)
r2_score(y_test,preds)
```

```python
##Random Forest
```

```python
from sklearn.ensemble import RandomForestRegressor
R_model = RandomForestRegressor()
R_model.fit(x_train,y_train)
preds = R_model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r = r2_score(y_test,preds)
print("R2score when we predict using Randomn forest is ",r)
```
