```python
f,axs = plt.subplots(10,10,figsize=(12, 12))
axs = axs.flatten()

for idx, ax in enumerate(axs):
    sns.kdeplot(data=train, x=f'f{idx}', 
                fill=True, 
                ax=ax)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_xlabel('')
    ax.set_ylabel('')
    ax.spines['left'].set_visible(False)
    ax.set_title(f'f{idx}', loc='right', weight='bold', fontsize=10)


f.tight_layout()
plt.show()
```

```python
X=train.iloc[:,:-1]
Y=train['loss']
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif

vif_X=pd.DataFrame()
vif_X['feature']=X.columns
vif_X['VIF_Factor']=[vif(X.values,i) for i in range(X.shape[1])]
vif_X.round(2)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X)
X_test=sc.transform(test)
y_train=Y
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
from sklearn.linear_model import LinearRegression
lr=LinearRegression()

lm_1=lr.fit(X_train,y_train)
pred_1=lm_1.predict(X_train)
pred_test=lm_1.predict(X_test)
rmse_1=mean_squared_error(y_train,pred_1,squared=False)
rmse_1
```

```python
output=pd.DataFrame({'id':test_1.id,'loss':pred_test})
output.to_csv('submission.csv', index=False)
```
