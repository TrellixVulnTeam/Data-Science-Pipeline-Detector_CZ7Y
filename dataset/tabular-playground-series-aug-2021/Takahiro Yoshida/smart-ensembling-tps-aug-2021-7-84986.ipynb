```python
sub1 = pd.read_csv('../input/tps-aug-2021-simple-weighted-ensemble/submission_weighted_ensemble_24.csv').sort_values('id')
sub2 = pd.read_csv('../input/blending-tool-tps-aug-2021/blend.csv/0.part').sort_values('id') 
sub3 = pd.read_csv('../input/84996to-improve-your-ranking/file1_7.84996_file2_7.84996_blend.csv').sort_values('id') 
sub4 = pd.read_csv('../input/k/yus002/blending-tool-tps-aug-2021/file1_7.84987_file2_7.84996_blend.csv').sort_values('id') 
```

```python
# Group data together
hist_data = [sub1.loss, sub2.loss, sub3.loss, sub4.loss]

group_labels = ['sub1', 'sub2', 'sub3', 'sub4']

# Create distplot with custom bin_size
fig = ff.create_distplot(hist_data, group_labels, bin_size=.2, show_hist=False, show_rug=False)
fig.show()
```

```python
data=np.corrcoef([sub1.loss, sub2.loss, sub3.loss, sub4.loss])
fig = px.imshow(data,
                x=group_labels,
                y=group_labels
               )
fig.show()
```

```python
sub = sub1.copy()
sub.loc[:, 'loss'] = (0.9 * sub1 + 0.1 * sub2 + 0.01)
sub.to_csv('submission.csv', index=False)
```

```python
# Group data together
hist_data = [sub1.loss, sub2.loss, sub3.loss, sub4.loss, sub.loss]

group_labels = ['sub1', 'sub2', 'sub3', 'sub4', 'sub']

# Create distplot with custom bin_size
fig = ff.create_distplot(hist_data, group_labels, bin_size=.2, show_hist=False, show_rug=False)
fig.show()
```
