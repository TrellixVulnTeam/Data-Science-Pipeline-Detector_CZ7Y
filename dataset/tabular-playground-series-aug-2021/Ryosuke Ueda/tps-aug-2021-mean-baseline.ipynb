# The Simple Baseline: Mean and Median

The public score (RMSE) are as follows.
* Mean: **7.98591**
* Median: 8.48468

When predicting a single value for all test IDs, the optimal solution is mean.
I also tried the median because the distribution of training and testing might be different.

```python
import pandas as pd
```

```python
train = pd.read_csv('../input/tabular-playground-series-aug-2021/train.csv')
test = pd.read_csv('../input/tabular-playground-series-aug-2021/test.csv')
```

```python
mean = train['loss'].mean()
median = train['loss'].median()

print('Mean:', mean)
print('Median:', median)
```

```python
sub_mean = pd.DataFrame({
    'id': test['id'],
    'loss': mean
})
sub_mean.to_csv('mean.csv', index=False)
```

```python
sub_median = pd.DataFrame({
    'id': test['id'],
    'loss': median
})
sub_median.to_csv('median.csv', index=False)
```
