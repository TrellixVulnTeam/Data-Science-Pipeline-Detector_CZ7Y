```python
#module
import numpy as np
import pandas as pd
from tpot import TPOTRegressor
```

```python
#data
train = pd.read_csv('/kaggle/input/tabular-playground-series-aug-2021/train.csv')
test = pd.read_csv('/kaggle/input/tabular-playground-series-aug-2021/test.csv')
sample_sub = pd.read_csv('../input/tabular-playground-series-aug-2021/sample_submission.csv')
```

```python
#preprocessing
y = train['loss']
train = train.drop(['id','loss'], axis=1)
X_train = train
X_test = test[X_train.columns]
```

```python
#model
tpot_r = TPOTRegressor(scoring='neg_mean_squared_error',
                     generations=5,
                     population_size=50,
                     random_state=42,
                     early_stop=20,
                     verbosity=2,
                     config_dict='TPOT cuML')
```

```python
#learning
tpot_r.fit(X_train,y)
```

```python
#predict
pred = tpot_r.predict(X_test)
```

```python
#submit
solution = pd.DataFrame({'id':sample_sub['id'], 'loss':pred})
solution.to_csv('tps_aug_tpot.csv', index = False)
```
