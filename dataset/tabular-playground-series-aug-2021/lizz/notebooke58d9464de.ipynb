```python
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['font.family'] = 'sans-serif'
# 解决负号'-'显示为方块的问题
plt.rcParams['axes.unicode_minus'] = False
```

```python
from sklearn.ensemble import RandomForestRegressor
import xgboost as xgb
import lightgbm as lgb
from sklearn.metrics import mean_squared_error, precision_score, recall_score, accuracy_score, roc_auc_score

# Hyperparameter tuning
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV, cross_val_score
```

```python
def evaluate_model_performance(df_features,df_target,model_estimators_lst):
    model_scores_lst = []
    for estimator in model_estimators_lst:
        model_score = cross_val_score(estimator, df_features, df_target, cv=3, n_jobs=-1, 
                                      scoring='neg_root_mean_squared_error',
                                     verbose=2)
        print('current model score is: {0}, mean score is: {1}'.format(model_score,model_score.mean()))
        model_scores_lst.append(model_score)
        
    # Dataframe to hold the results
    model_comparison = pd.DataFrame({'model': ['Random Forest', 
                                               'XGBoost',
                                               'LightGBM'],
                                 'score': [model_scores_lst[0].mean(), model_scores_lst[1].mean(), 
                                         model_scores_lst[2].mean()]})

    # Horizontal bar chart of test mae
    model_comparison.sort_values('score', ascending = False).plot(x = 'model', y = 'score', kind = 'barh',
                                                           color = 'red', edgecolor = 'black')

    # Plot formatting
#     x = np.array(list(model_comparison.sort_values('score', ascending = False)['score']))
#     y = np.array([1, 2, 3])
#     for a, b in zip(x, y):
#         plt.text(a+0.05, b, '%.0f' % a, ha='center', va= 'center',fontsize=7)
        
    plt.ylabel(''); plt.yticks(size = 14); plt.xlabel('cross validation score'); plt.xticks(size = 14)
    plt.title('Model Comparison on train set', size = 20);
```
