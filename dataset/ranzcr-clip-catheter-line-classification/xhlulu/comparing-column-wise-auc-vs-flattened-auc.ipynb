```python
import os
import gc
import numpy as np
import pandas as pd
from sklearn.metrics import roc_auc_score
```

```python
train = pd.read_csv('../input/ranzcr-clip-catheter-line-classification/train.csv')
```

```python
y_true = train.iloc[:, 1:-1]
```

```python
y_pred = pd.concat([y_true.mean()] * y_true.shape[0], axis=1).T
```

Roc AUC when flattening the vectors:

```python
y_true_flat = y_true.values.reshape(-1)
y_pred_flat = y_pred.values.reshape(-1)

roc_auc_score(y_true_flat, y_pred_flat)
```

ROC AUC when calculating each column separately:

```python
for col in y_true.columns:
    print(col, roc_auc_score(y_true[col], y_pred[col]))
```

ROC AUC as mean over columns:

```python
roc_auc_score(y_true, y_pred)
```
