# RANZCR CLiP - Catheter and Line Position Challenge

```python
import pandas as pd
import os
import json
import cv2
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (30,10)
```

```python
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
```

```python
base_folder='../input/ranzcr-clip-catheter-line-classification'
image_folder='../input/ranzcr-clip-catheter-line-classification/train'
```

```python
df_train=pd.read_csv(os.path.join(base_folder, 'train.csv'))
df_mask=pd.read_csv(os.path.join(base_folder, 'train_annotations.csv'))
```

```python
df_train.head()
```

```python
df_mask.head()
```

```python
df_ETT_ab=df_mask[df_mask['label']=='ETT - Abnormal'].reset_index(drop=True)
```

```python
df_ETT_ab
```

```python
def showing_img(index):
    print (df_ETT_ab.iloc[index].StudyInstanceUID)
    data=df_ETT_ab.iloc[index]['data']
    coors=json.loads(data)
    images=cv2.imread(os.path.join(image_folder, df_ETT_ab.iloc[index]['StudyInstanceUID']+'.jpg'),0)
    images=cv2.cvtColor(clahe.apply(images),cv2.COLOR_GRAY2RGB)

    images_ori=images.copy()
    h,w=images.shape[0:2]
    blue = (0, 0, 255)
    for n, j in enumerate(coors):
        if n != len(coors)-1:
            print (coors[n])
            cv2.line(images, tuple(coors[n]), tuple(coors[n+1]), blue, 3)
    
    plt.subplot(121)
    plt.imshow(images_ori)
    plt.subplot(122)
    plt.imshow(images)
```

Before we start, please check out domain knowledge from dr. Konya's great article (https://www.kaggle.com/c/ranzcr-clip-catheter-line-classification/discussion/204790)

# The abnormal tip position of endotracheal tube

```python
showing_img(2)
```

No visualization of the tip of endotracheal tube near tracheal bifurcation
