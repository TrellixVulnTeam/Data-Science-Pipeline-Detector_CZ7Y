# **Let's Check if we have some samples with contradicting features**

```python
ETT_cols = ['ETT - Abnormal', 'ETT - Borderline', 'ETT - Normal']
NGT_cols = ['NGT - Abnormal', 'NGT - Borderline', 'NGT - Incompletely Imaged', 'NGT - Normal']
CVC_cols = ['CVC - Abnormal', 'CVC - Borderline', 'CVC - Normal']
```

```python
train_df[ETT_cols].sum(axis=1).value_counts()
```

```python
train_df[NGT_cols].sum(axis=1).value_counts()
```

> ### **The samples with sum as 2 have '1' for more than one column which doesn't seem correct, Let's explore a bit more**

```python
train_df[train_df[NGT_cols].sum(axis=1) == 2]
```

> ### **We can see that some samples have been classified into both NGT-Normal and NGT-Abnormal which definitely can't be possible**

```python
train_df[CVC_cols].sum(axis=1).value_counts()
```

> ### Again we have samples with '1' for more than one column

```python
train_df[train_df[CVC_cols].sum(axis=1) == 2]
```

> ### **There are many samples where we have more than 1 correct class which again looks like a case of misclassification**

```python
train_df[train_df[CVC_cols].sum(axis=1) == 3]
```

> ### **These samples definitely have wrong labels as they are classified in all the 3 categories i.e. CVC-Abnormal, CVC-Normal and CVC-Borderline**
