```python
from tensorflow.keras.models import Model
import tensorflow as tf
import pandas as pd
import numpy as np
import keras
```

```python
model = keras.models.load_model('../input/inceptionnetv3-on-rancr/model.h5')
```

```python
data_path = "../input/ranzcr-clip-catheter-line-classification"
preds = []
ss = pd.read_csv(data_path+'/sample_submission.csv')
```

```python
for image in ss.StudyInstanceUID:
    img = tf.keras.preprocessing.image.load_img(data_path+'/test/' + image + '.jpg')
    img = tf.keras.preprocessing.image.img_to_array(img)
    img = tf.keras.preprocessing.image.smart_resize(img, (256,256))
    img = tf.reshape(img, (-1, 256,256, 3))
    prediction = model.predict(img/255)
    prediction = prediction.reshape(11)
    preds.append(prediction)
```

```python
submission = pd.DataFrame(preds, columns=[
 'ETT - Abnormal',
 'ETT - Borderline',
 'ETT - Normal',
 'NGT - Abnormal',
 'NGT - Borderline',
 'NGT - Incompletely Imaged',
 'NGT - Normal',
 'CVC - Abnormal',
 'CVC - Borderline',
 'CVC - Normal',
 'Swan Ganz Catheter Present'])
```

```python
submission['StudyInstanceUID'] = ss.StudyInstanceUID

submission = submission[['StudyInstanceUID','ETT - Abnormal','ETT - Borderline','ETT - Normal','NGT - Abnormal','NGT - Borderline','NGT - Incompletely Imaged','NGT - Normal','CVC - Abnormal','CVC - Borderline','CVC - Normal','Swan Ganz Catheter Present']]

submission.to_csv('./submission.csv', index=False)
```
