# Resizing image for faster training

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import matplotlib.image as immg
from pathlib import Path
import os
import gc
import cv2
from tqdm.notebook import tqdm
import zipfile
import io
from sklearn.decomposition import PCA
```

```python
img_dir = Path('../input/ranzcr-clip-catheter-line-classification/train')
```

```python
df = pd.read_csv("../input/ranzcr-clip-catheter-line-classification/train.csv")
```

```python
files = [str(x)+'.jpg' for x in df.StudyInstanceUID.tolist()]
```

## So I will reduce and save to 512x512 of its size

```python
OUT_TRAIN = 'trainXray_512.zip'
```

```python
x_tot,x2_tot = [],[]
with zipfile.ZipFile(OUT_TRAIN, 'w') as img_out:
    for idx in tqdm(range(len(files))):
        name = str(files[idx])[:-4]
        img = cv2.imread(str(img_dir/files[idx]), 1) 
        img = cv2.resize(img, (512, 512))
        x_tot.append((img/255.0).mean())
        x2_tot.append(((img/255.0)**2).mean()) 
        
        img = cv2.imencode('.jpg',img)[1]
        img_out.writestr(name + '.jpg', img)
```

## Image stats

```python
img_avr =  np.array(x_tot).mean()
img_std =  np.sqrt(np.array(x2_tot).mean() - img_avr**2)
print('mean:',img_avr, ', std:', img_std)
```
