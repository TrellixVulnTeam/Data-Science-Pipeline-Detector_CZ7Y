```python
import pandas as pd

fold5_1 = pd.read_csv('../input/ranzcr-prob/prob1.csv')
fold5_2 = pd.read_csv('../input/ranzcr-prob/prob2.csv')
fold5_3 = pd.read_csv('../input/ranzcr-prob/prob3.csv')
fold5_4 = pd.read_csv('../input/ranzcr-prob/prob4.csv')
fold5_5 = pd.read_csv('../input/ranzcr-prob/prob5.csv')
columns = fold5_1.columns
```

```python
P = [0.4, 0.4, 0, 0.1, 0.1]
for i in columns[1:]:
    fold5_1[i] = fold5_1[i] * P[0]  + fold5_2[i] * P[1] + fold5_3[i] * P[2] + \
                 fold5_4[i] * P[3] + fold5_5[i] * P[4]
```

```python
fold5_1.to_csv('sub.csv', index=False)
```
