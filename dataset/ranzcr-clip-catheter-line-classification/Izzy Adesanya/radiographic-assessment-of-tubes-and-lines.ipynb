# Endotracheal Tubes (ETT) – Normal Position


1. Gold standard for determining placement in airway is end-tidal CO2

* Limited ability to detect some complications

* Physical exam maneuvers also may be helpful

2. Radiographic assessment of tube position based on visualizing radioopaque line on ETT

3. ETT position dependent on neck flexion and extension:
   * Neck flexion (mandible @ T1-2): ETT tip can descend up to 2 cm
   * Neck extension (mandible @ C4): ETT tip can ascend up to 2 cm
   
4. Acceptable positioning:
  * Proximal tube tip: level of medial heads of the clavicle
  * Too proximal – risk upper airway/vocal cord injury
  * Distal tube tip: 2 cm above carina
  * Too distal – mainstem bronchus intubation with neck flexion
  * Tracheal cuff should fill tracheal wall (not bulge)

# Endotracheal Tubes – Abnormal Position

1. Potential Complications:
   * Increased risk of spontaneous
   * extubation
   * Aspiration
   * Vocal cord injury
   * Gastric distension
   * Left lobe (R main stem bronchus) or right upper lobe (R inferior intermediate bronchus) collapse/atelectasis
   * Tension pneumothorax

# Tracheostomy Tubes – Normal Position

1. Similar distal tip positioning rules as ETTs

2. Tip position should NOT vary with neck flexion/extension

3. At least 2/3 of “smooth” portion should lie within the trachea

# Tracheostomy Tubes – Complications

Potential complications
* Tracheal injury
    * Subcutaneous emphysema/pneumomediastinum
* Pneumothorax
* Hemorrhage
* Late complications:
    * Tracheal stenosis
    * Tracheomalacia
    * Vascular erosions
    * Tracheobronchial fistula

# Central Venous Catheters – Normal Position


1. Central venous catheter may be central (IJ, subclavian) or peripheral (PICC)

2. Catheter tip should terminate in SVC or cavo-atrial junction
  * “OK to use” varies with intended function

3. Placement borders:
  * Brachiocephalic/IJ junction – 1st anterior intercostal space
    * Last venous valve
  * Cavo-atrial junction – inferior border of right bronchus intermedius (+2.5 cm below)
    * Prevent cardiac chamber insertion

# Central Venous Catheters – Abnormal Position


Positioning errors:

* Ipsilateral IJ
* Contralateral brachiocephalic vein
* Right atrium/ventricle
  * Increased risk of arrhythmias
  * Reduced dilution of administered medications
* Azygous vein*
* Internal thoracic vein*
