```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("../input/ranzcr-clip-catheter-line-classification/train.csv")
df.head()
```

## PatientID

```python
print(len(df.PatientID.value_counts()))
```

```python
fig = px.bar(df.PatientID.value_counts(), title="Unique Patient Count")
fig.show()
```

## Label analys

```python
len(df.columns[1:-1])
```

```python
label = df.columns[1:-1]
sum_label = df[label].values.sum(1)
px.bar(["Multi label", "Single label"], [sum(sum_label>1),sum(sum_label==1)], title="Single label - Multi label")
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objs as go

fig = make_subplots(rows=5, cols=3)

traces = [
    go.Bar(
        x=[0, 1], 
        y=[
            len(df[df[col]==0]),
            len(df[df[col]==1])
        ], 
        name=col,
        text = [
            str(round(100 * len(df[df[col]==0]) / len(df), 2)) + '%',
            str(round(100 * len(df[df[col]==1]) / len(df), 2)) + '%'
        ],
        textposition='auto'
    ) for col in label
]

for i in range(len(traces)):
    fig.append_trace(traces[i], (i // 3) + 1, (i % 3)  +1)

fig.update_layout(
    title_text='Train columns',
    height=1200,
    width=1000
)

fig.show()
```

```python
x = df.drop(['StudyInstanceUID', 'PatientID'], axis=1).sum(axis=0).sort_values().reset_index()
x.columns = ['column', 'nonzero_records']

fig = px.bar(
    x, 
    x='nonzero_records', 
    y='column', 
    orientation='h', 
    title='Columns and non zero samples', 
    height=800, 
    width=800
)

fig.show()
```

## Size of all image

```python
import os
from pathlib import Path
size_arr = []
TRAIN_FOLDER = Path("../input/ranzcr-clip-catheter-line-classification/train")
for idx, img in enumerate(os.listdir(TRAIN_FOLDER)):
    test_image = cv.imread(os.path.join(TRAIN_FOLDER, img))
    size_arr.append(test_image.shape)
    if idx > 1000:
        break
```

```python
df_size = pd.DataFrame({"sizeimg": size_arr})
df_size.sizeimg = df_size.sizeimg.map(str)
fig = px.bar(df_size.sizeimg.value_counts().tolist(),  title="img size")
fig.show()
```
