```python
import pandas
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import random
```

```python
PATH = "../input/ranzcr-clip-catheter-line-classification/"
TRAIN_PATH = PATH + "train/"
```

## Class Distribution

```python
labels = pandas.read_csv(PATH + "train.csv")
print("Number of images in dataset:", len(labels))
categoryTotals = np.sum(labels.drop(["StudyInstanceUID", "PatientID"], axis=1))
print("Labels:\n", categoryTotals)
```

## Image Sizes

```python
widths = []
heights = []

for imgName in os.listdir(TRAIN_PATH):
    img = Image.open(TRAIN_PATH + imgName)
    w, h = img.size
    widths.append(w)
    heights.append(h)
```

```python
plt.suptitle("Image Widths")
plt.hist(widths, bins=10)
plt.show()
```

```python
plt.suptitle("Image Heights")
plt.hist(heights, bins=10)
plt.show()
```

## Sample Images By Class

```python
def displaySampleClassImages(className):
    numImgs = 6
    ids = labels[labels[className]==1]["StudyInstanceUID"] #ids belonging to class
    sampleImgIds = ids.sample(numImgs).to_numpy()
    
    fig = plt.figure(figsize=(12,8))
    cols = 3
    rows = 2
    
    for i in range(1, numImgs+1):
        fig.add_subplot(rows, cols, i)
        #image needs converting to display as black and white
        img = Image.open(TRAIN_PATH + sampleImgIds[i-1] + ".jpg").convert("RGB")
        plt.imshow(img)
    
    plt.show()
```
