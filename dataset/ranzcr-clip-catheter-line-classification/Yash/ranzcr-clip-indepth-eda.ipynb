```python
import pandas as pd
import os
import cv2
import plotly.graph_objects as go
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
BASE_DIR = "../input/ranzcr-clip-catheter-line-classification/"
df_train = pd.read_csv(os.path.join(BASE_DIR, "train.csv"), index_col=0)
df_train.head()
```

```python
df_train.index.nunique()
```

```python
data = df_train.iloc[:, :-1].sum()

fig = go.Figure(data=[
    go.Pie(labels=data.index, values=data.values)
])

fig.update_layout(title='Data distribution')
fig.show()
```

```python
import random
fig = go.Figure()

for col in data.index:
    
    random_number = random.randint(0, 16777215)
    hex_number = str(hex(random_number))
    hex_number = '#' + hex_number[2:]
    fig.add_trace(go.Histogram(
        x=df_train[col],
        histnorm='percent',
        name=col,
        marker_color=hex_number,
        opacity=0.75
    ))
fig.show()
```

```python
annot = pd.read_csv(os.path.join(BASE_DIR, "train_annotations.csv"))
annot.head()
```

```python
import numpy as np
import ast
plt.show(block=False)
def image_plot(images, labels, datas):
    plt.figure(figsize=(16, 8))
    for ind, (image, label, data) in enumerate(zip(images, labels, datas)):
        path = os.path.join(BASE_DIR, "train", image + ".jpg")
        req_image = cv2.imread(path)
        req_image = cv2.cvtColor(req_image, cv2.COLOR_BGR2RGB)
        data = np.array(ast.literal_eval(data))
        plt.subplot(2, 2, ind+1)
        plt.imshow(req_image)
        plt.title(f"Class: {label}", fontsize=12)
        plt.subplot(2, 2, ind+3)
        plt.imshow(req_image)
        plt.plot(data[:, 0], data[:, 1])
        plt.title(f"Class: {label}", fontsize=12)
        plt.axis("off")
    
```

```python
for col in data.index:
    df = annot.loc[annot['label'] == col]

    image_plot(df.sample(2)['StudyInstanceUID'].values, df.sample(2)['label'].values, df.sample(2)['data'])
```
