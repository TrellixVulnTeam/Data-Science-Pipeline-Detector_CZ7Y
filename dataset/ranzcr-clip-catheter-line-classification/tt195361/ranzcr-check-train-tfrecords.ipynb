## Check Train TFRecords

This notebook checks the training TFRecords provided for the RANZCR CLiP competition. There are 2 points to check:
1. Items in the TFrecord files: PatientID is not provided.
2. Number of items in each TFRecord files: the last file '15-1881.tfrec' has 1868 items. Others have 1881 items which are shown by the file name 'xx-1881.tfrec'.

```python
import tensorflow as tf

print(tf.__version__)
```

```python
RANZER_DIR = '../input/ranzcr-clip-catheter-line-classification/'
tfrec_file_names = tf.io.gfile.glob(RANZER_DIR + 'train_tfrecords/*.tfrec')
```

### 1. Items in The TFrecord Files

```python
dataset = tf.data.TFRecordDataset(tfrec_file_names)
raw_record = next(iter(dataset.take(1)))
example = tf.train.Example()
example.ParseFromString(raw_record.numpy())
feature_list = sorted([f for f in example.features.feature])

for f in feature_list:
    print(f)
```

### 2. Number of Items in Each TFRecord Files

```python
import os

for file_name in sorted(tfrec_file_names):
    dataset = tf.data.TFRecordDataset(file_name)
    item_count = 0
    for item in iter(dataset):
        item_count += 1
    base_name = os.path.basename(file_name)
    print("{0}: {1}".format(base_name, item_count))
```
