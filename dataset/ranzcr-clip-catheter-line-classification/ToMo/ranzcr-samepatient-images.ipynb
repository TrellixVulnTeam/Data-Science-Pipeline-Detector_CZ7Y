```python
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
plt.style.use("seaborn-white")
import ast

BASE_DIR = "../input/ranzcr-clip-catheter-line-classification"
TRAIN_DIR = "../input/ranzcr-clip-catheter-line-classification/train"
TEST_DIR = "../input/ranzcr-clip-catheter-line-classification/test"

df = pd.read_csv(BASE_DIR + "/train.csv")
an = pd.read_csv(BASE_DIR + "/train_annotations.csv")
LABELS = [
    'ETT - Abnormal', 'ETT - Borderline', 'ETT - Normal',
    'NGT - Abnormal', 'NGT - Borderline', 'NGT - Incompletely Imaged', 'NGT - Normal', 
    'CVC - Abnormal', 'CVC - Borderline', 'CVC - Normal',
    'Swan Ganz Catheter Present'
]

def show_images(studyuid):
    path = TRAIN_DIR + "/" + studyuid + ".jpg"
    image = cv2.imread(path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    annotations = an.loc[an["StudyInstanceUID"] == studyuid]
    _, ax = plt.subplots(1, 2, figsize = (16, 6))
    ax[0].imshow(image)
    ax[1].imshow(image)
    for i in range(annotations.shape[0]):
        points = np.array(ast.literal_eval(annotations["data"].values[i]))
        ax[1].scatter(points[:, 0], points[:, 1], label = annotations["label"].values[i])
    plt.legend(bbox_to_anchor = (1.5, 1))
    plt.show()
```

```python
dummy = []
for studyuid in df.loc[df["PatientID"] == "ab02c52b6"]["StudyInstanceUID"].values:
    num = an.loc[an["StudyInstanceUID"] == studyuid].shape[0]
    if num > 0:
        dummy.append([studyuid, num])
print(len(dummy))
```

```python
dummy = pd.DataFrame(np.stack(dummy, axis = 0), columns = ["studyuid", "num"]).sort_values(by = "num")
dummy
```

```python
ids = dummy["studyuid"].values
for i in range(len(ids)):
    show_images(ids[i])
```
