# Load train and test as DataFrames

```python
train = pd.read_csv('../input/ranzcr-clip-catheter-line-classification/train.csv')
test = pd.read_csv('../input/ranzcr-clip-catheter-line-classification/sample_submission.csv')
print(train.shape)
print(test.shape)
train.head(10)
```

# Check distribution of labels in train

```python
train.mean()
```

```python
TARGET = []
PREDICTION = []
for target in train.columns[1:-1]:
    TARGET.extend( train[target].values )
    PREDICTION.extend( list(np.ones(train.shape[0])*train[target].mean()) )

print( 'Baseline Score:', roc_auc_score( TARGET, PREDICTION ) )
```

# Write train target mean average to each test label

```python
for target in train.columns[1:-1]:
    print(target)
    test[target] = train[target].mean()
    
test.head()
```
