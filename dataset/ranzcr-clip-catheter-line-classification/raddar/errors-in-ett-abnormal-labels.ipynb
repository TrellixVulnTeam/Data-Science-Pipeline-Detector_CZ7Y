## Abnormal ETT - are they labelled correctly?

As there are only 30 `ETT - Abnormal` cases for which we annotations - let's visualize all and decide if all of them are correctly labelled.

If there are at least one wrong label, it could have huge impact on AUC scores for `ETT - Abnormal` class

```python
import ast
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

train = pd.read_csv('../input/ranzcr-clip-catheter-line-classification/train.csv')
ann = pd.read_csv('../input/ranzcr-clip-catheter-line-classification/train_annotations.csv')
ann = ann.loc[ann.label == 'ETT - Abnormal']
print(ann.shape)
```

```python
def plot_xray(StudyInstanceUID):
    """
    intubation points as green
    """
    img = cv2.imread('../input/ranzcr-clip-catheter-line-classification/train/'+StudyInstanceUID+'.jpg')
    intubation = ast.literal_eval(ann.loc[(ann.StudyInstanceUID==StudyInstanceUID),'data'].values[0])
    for point in intubation:
        img = cv2.circle(img, tuple(point), 10, (0,255,0), 10)
    plt.figure(figsize=(8,8))
    print(f'StudyInstanceUID {StudyInstanceUID}')
    plt.imshow(img)
```

First train instance seems to be ok:

```python
plot_xray(ann.StudyInstanceUID[ann.index[0]])
```

However, the very second instance is not ok! it should be labelled as CVC - Abnormal!

```python
plot_xray(ann.StudyInstanceUID[ann.index[1]])
```

I looked through the rest of the cases - all other 28 cases seem to be fine.

So, during loading train data make sure to correct the case above,

this should do the trick:



```python
train.loc[train.StudyInstanceUID=='1.2.826.0.1.3680043.8.498.93345761486297843389996628528592497280', 'ETT - Abnormal'] = 0
train.loc[train.StudyInstanceUID=='1.2.826.0.1.3680043.8.498.93345761486297843389996628528592497280', 'CVC - Abnormal'] = 1
ann.loc[4344, 'label'] = 'CVC - Abnormal'
```
