```python
# importing modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

# Graphics in retina format are more sharp and legible
%config InlineBackend.figure_format = 'retina'
```

```python
# reading training data.
data_train = pd.read_csv("/kaggle/input/santa-workshop-tour-2019/family_data.csv")
```

```python
print(data_train.info())
```

Okay..! So all the columns present in our data is having **integer values** and with no **Nan**. Looks good.. :)

Before, exploring the type of data present in every columns let's have some look at the dataset at once.

```python
data_train.head(10)
```

```python
print("Column name - Unique values")
for col in data_train.columns:
    print(col, " - ", len(data_train[col].value_counts()))
```

Now Let's visualize every column of the dataset.

we don't need to visualize the column **family_id** as it is just the serial number of every rows.

```python
# Visualize 'choice_0'
data_train['choice_0'].hist(figsize=(10, 4),bins=100);
```

```python
# Visualize 'choice_1'
data_train['choice_1'].hist(figsize=(10, 4),bins=100);
```

```python
# Visualize 'choice_2'
data_train['choice_2'].hist(figsize=(10, 4),bins=100);
```

It looks like every `choice_n` column is having almost similar disturbution of data. So, let's visualize them together and observe it.

```python
# Visualize all choice features
features = data_train.columns[1:-1]

for feature in features:
    data_train[feature].hist(figsize=(70, 50),bins=100);
```

The summarized form of visualization plot for 'choice_n` column shows that the value '0' is present in large amount in every column and also other values are disturbuted in similar pattern.


```python
# Visualizing 'n_people' column
data_train['n_people'].hist(figsize=(10, 4),bins=7);
```

In column, 'n_people' 4 is present in large quantity in comparison to other. <br>
This shows that the families like to come on 4 (i.e. 24 Dec) the most and this trend starts from 2 (i.e. 22 Dec) and this trend starts to decrease after 4.
