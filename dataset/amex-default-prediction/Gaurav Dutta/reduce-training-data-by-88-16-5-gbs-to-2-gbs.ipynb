```python
train.S_2=pd.to_datetime(train.S_2, infer_datetime_format=True)
train.info()
```

```python
from fast_ml.feature_engineering import FeatureEngineering_DateTime
dt_fe = FeatureEngineering_DateTime()
dt_fe.fit(train, datetime_variables=['S_2'])
train = dt_fe.transform(train)
train.head()
```

```python
nunique_train=train.nunique().reset_index()
remove_col=nunique_train[(nunique_train[0]==len(train)) | (nunique_train[0]==0) | (nunique_train[0]==1) ]['index'].tolist()
remove_col
```

```python
pd.set_option('display.max_columns', None)
train=train.drop(remove_col,axis=1)
train.head()
```

```python
total_col=train.columns.tolist()
total_col
```

```python
train=train.drop(['S_2'],axis=1)
train[['S_2:year',
 'S_2:quarter',
 'S_2:month',
 'S_2:day',
 'S_2:day_of_week',
 'S_2:day_of_year',
 'S_2:weekofyear',
 'S_2:is_month_end',
 'S_2:is_month_start',
 'S_2:is_quarter_end',
 'S_2:is_quarter_start',
 'S_2:is_year_end',
 'S_2:is_year_start',
 'S_2:is_weekend']]=train[['S_2:year',
 'S_2:quarter',
 'S_2:month',
 'S_2:day',
 'S_2:day_of_week',
 'S_2:day_of_year',
 'S_2:weekofyear',
 'S_2:is_month_end',
 'S_2:is_month_start',
 'S_2:is_quarter_end',
 'S_2:is_quarter_start',
 'S_2:is_year_end',
 'S_2:is_year_start',
 'S_2:is_weekend']].astype('int8')
train.shape
```

```python
train_y=pd.read_csv('../input/amex-default-prediction/train_labels.csv',dtype={'target':np.dtype('int8')})
train_y=train_y[['target']]
train_y.head()
```

```python
train['target']=train_y['target']
del train_y
train.target=train.target.astype('float16')
train.head()
```
