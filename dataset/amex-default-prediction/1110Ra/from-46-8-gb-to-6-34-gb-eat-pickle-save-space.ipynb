All float64 and int64 columns are converted to float16 and int8. In order to read the files, you can use **pd.read_pickle()**. This method will save loading time and reduce RAM usage dramatically. 

#### Reduce Dataset Size and Create Pickle format

```python
import pandas as pd
import gc
```

```python
train_data = pd.read_csv('../input/amex-default-prediction/train_data.csv', index_col='customer_ID')
```

```python
train_data.info()
```

```python
train_data.dtypes.value_counts()
```

```python
float64_columns = train_data.columns[train_data.dtypes == 'float64'].values
object_columns = train_data.columns[train_data.dtypes == 'O'].values
int64_columns = train_data.columns[train_data.dtypes == 'int64'].values
```

```python
float64_columns
```

```python
object_columns
```

```python
int64_columns
```

```python
train_data[float64_columns] = train_data[float64_columns].astype('float16')
train_data[int64_columns] = train_data[int64_columns].astype('int8')
```

```python
train_data.info()
```

```python
train_data.to_pickle('train_data.pkl')
```

```python
del train_data
gc.collect()
```

```python
test_data = pd.read_csv('../input/amex-default-prediction/test_data.csv', index_col='customer_ID')
```

```python
test_data.info()
```

```python
test_data[float64_columns] = test_data[float64_columns].astype('float16')
test_data[int64_columns] = test_data[int64_columns].astype('int8')
```

```python
test_data.info()
```

```python
test_data.to_pickle('test_data.pkl')
```

```python
del test_data
gc.collect()
```
