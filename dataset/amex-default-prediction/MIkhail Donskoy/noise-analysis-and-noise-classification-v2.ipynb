#### Such features can be seen on the charts below.

```python
fig = plt.figure(figsize=(20,5))
plt.subplot(1, 3, 1)
sns.histplot(data=train, x="B_31", hue="target", bins=400)
plt.subplot(1, 3, 2)
sns.histplot(data=train, x="B_32", hue="target", bins=400)
plt.subplot(1, 3, 3)
sns.histplot(data=train, x="B_33", hue="target", bins=400)
plt.show()
```

```python
def plot_noise(train,col, treshold, bins = 400):
    fig = plt.figure(figsize=(20,5))
    plt.subplot(1, 3, 1)
    sns.histplot(data=train, x=col, hue="target", bins=bins)
    plt.subplot(1, 3, 2)
    sns.histplot(data=train[train[col] < treshold], x=col, hue="target", bins=bins)
    plt.subplot(1, 3, 3)
    sns.histplot(data=train[train[col] > treshold], x=col, hue="target", bins=bins)
    plt.show() 
```
