```python
# lets read train labels
# as the file size is not so huge thus no need to use data types or chunksize
start = time.time()
train_labels_df = pd.read_csv('/kaggle/input/amex-default-prediction/train_labels.csv')  # gives TextFileReader
print('time taken to read',time.time()-start)
train_labels_df.shape
```

```python
# merge the train data with train labels
start = time.time()
train_df = pd.merge(df,train_labels_df,left_on='customer_ID',right_on='customer_ID',how='left')
print('time taken to read',time.time()-start)
train_df.shape
```
