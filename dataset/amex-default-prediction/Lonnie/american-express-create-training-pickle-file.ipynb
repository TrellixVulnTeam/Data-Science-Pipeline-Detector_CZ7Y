## American Express: Create Training Pickle File

```python
import numpy as np
import pandas as pd
import gc
```

```python
%%time
train_data = pd.DataFrame()
with pd.read_csv('/kaggle/input/amex-default-prediction/train_data.csv', chunksize=10**5) as reader:
    for counter, chunk in enumerate(reader):
        for column in chunk.columns:
            if str(chunk[column].dtype) == "float64":
                chunk[column] = chunk[column].astype(np.float16)
            if str(chunk[column].dtype) == "int64":
                chunk[column] = chunk[column].astype(np.int16)
        train_data = pd.concat([train_data, chunk])
        gc.collect()
```

```python
train_data.to_pickle("train_data.pkl")
```

```python
train_labels = pd.read_csv("/kaggle/input/amex-default-prediction/train_labels.csv")
train_labels.to_pickle("train_labels.pkl")
```
