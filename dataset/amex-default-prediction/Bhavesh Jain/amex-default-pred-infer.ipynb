## Prediction using model trained with LGBM

[Training Notebook](https://www.kaggle.com/code/bhavesjain/amex-default-pred-train?scriptVersionId=99324580)

```python
import pandas as pd
import numpy as np
from tqdm import tqdm 
import os

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler


import xgboost as xgb
import lightgbm as lgb
```

```python
df = pd.read_csv("../input/d/datasets/bhavesjain/amex-default-prediction/test-last-rows.csv")
df.fillna(df.mode().iloc[0],inplace=True)
```

```python
scaler = MinMaxScaler()
```

```python
x_cols = [col for col in df.columns[2:] if col not in ['D_63','D_64']]
```

```python
df[x_cols] = scaler.fit_transform(df[x_cols])
```

```python
# model = xgb.XGBClassifier()
# model.load_model("../input/amex-default-pred-train/model.json")
# pred = model.predict_proba(df[x_cols])[:,1]
```

```python
model = lgb.Booster(model_file='../input/amex-default-pred-train/lgb_model.json')
pred = model.predict(df[x_cols])
```

```python
df["prediction"] = pred
```

```python
submission_df = df[["customer_ID","prediction"]]
```
