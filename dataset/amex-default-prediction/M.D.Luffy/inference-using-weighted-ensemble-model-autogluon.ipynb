```python
predictor = TabularPredictor.load('../input/notebook832ca68bb9/agModels-predictClass/')
```

```python
y_pred = []
test_df = test_df.drop('customer_ID', axis=1)
for i in range(0, test_df.shape[0], 100000):
    y_pred.extend(predictor.predict_proba(test_df.iloc[i:i+100000])[1].to_list())
```
