# Quick overview

Iam using the parquet data of Sanskar

```python
%%time
train = pd.read_parquet('../input/amex-parquet/train_data.parquet')
train_labels=pd.read_csv('../input/amex-default-prediction/train_labels.csv')
```

```python
# missing table function
def missing_table(df):
    missing_df=pd.DataFrame({'Share in %':round((df.isnull().sum()/len(df))*100,1).values},index=[list(df.columns)])
    missing_df=missing_df.sort_values(by='Share in %',ascending=False)
    return missing_df

# show top 5 missing variables
missing_table(train).head(5)
```

```python
# check types of columns
train.dtypes.values
```

```python
# look up mean values for targets in train_labels
print(f'{round((train_labels.target.mean())*100,1)} % of customers default')
```

```python
# merge train with targets
temp=pd.merge(train[['customer_ID']],train_labels,on='customer_ID',how='left')
del train
```

```python
# frequencys and target distribution
temp1=pd.DataFrame(temp.customer_ID.value_counts()).reset_index().rename({'index':'customer_ID','customer_ID':'cust_bins'},axis=1)
temp2=pd.merge(temp1,train_labels,on='customer_ID',how='right')
temp2.groupby('cust_bins').target.agg(customer_amount_in_bins='count',target_share_in_bins='mean')
```
