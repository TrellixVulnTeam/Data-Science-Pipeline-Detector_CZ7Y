```python
%%time
train = pd.read_feather('../input/amexfeather/train_data.ftr')
#test = pd.read_feather('../input/amexfeather/test_data.ftr')
#sample_submission = pd.read_csv('../input/amex-default-prediction/sample_submission.csv')
```

```python
print(train.shape)
#print(test.shape)

train.head()
```

```python
x = train.loc[:, 'P_2':'D_145'].describe()
```

```python
x= x.T
```

```python
mean = []
std = []
for col in train.columns: #included target column in the string , so we will only consider mean[:-1] for our consideration
    if 'int' in str(train[col].dtypes) or 'float' in str(train[col].dtypes):
        #used .astype('float64') to avoid memory overflow and getting 'inf' as mean
        mean.append(np.nanmean(train[col].astype('float64'), axis = 0))
        std.append(np.nanstd(train[col].astype('float64'), axis = 0))
```

```python
x['mean'] = mean[:-1]
x['std'] = std[:-1]
```

```python
import gc
gc.collect()
```

```python
x.style.bar(subset=['mean'], color='#205ff2')\
                            .background_gradient(subset=['std'], cmap='Reds')\
                            .background_gradient(subset=['50%'], cmap='coolwarm')
```
