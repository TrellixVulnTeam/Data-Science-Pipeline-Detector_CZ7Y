# Get Min - Max values for float datatype columns

```python
# Dictionary to store values during import
min_max_dict = dict()
for x in float_cols:
    min_max_dict[x] = dict()
    min_max_dict[x]['min'] = None
    min_max_dict[x]['max'] = None
```

```python
# Function to get min and max values in each columns of given dataframe
def get_min_max(df):
    for c in df.columns:
        values = df[c].dropna().values
        min_value = None if len(values) == 0 else np.min(values)
        max_value = None if len(values) == 0 else np.max(values)
        if min_max_dict[c]['min'] != None:
            if min_value < min_max_dict[c]['min']:
                min_max_dict[c]['min'] = min_value
        else:
            min_max_dict[c]['min'] = min_value
        if min_max_dict[c]['max'] != None:
            if max_value > min_max_dict[c]['max']:
                min_max_dict[c]['max'] = max_value
        else:
            min_max_dict[c]['max'] = max_value
```

```python
# Importing the dataset in small chunks
chunksize = 1000000 #rows subset from large dataframe
for chunk in pd.read_csv('../input/amex-default-prediction/train_data.csv',usecols=float_cols, chunksize=chunksize):
    get_min_max(chunk)
```
