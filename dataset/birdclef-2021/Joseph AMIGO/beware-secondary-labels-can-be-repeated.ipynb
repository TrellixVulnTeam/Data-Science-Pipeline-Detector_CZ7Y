I thought there was something with secondary labels happening like repeatitions. I was right!
Hope it saves you time!

```python
import pandas as pd
meta_data_df = pd.read_csv('../input/birdclef-2021/train_metadata.csv', index_col="filename")
secondary_labels = meta_data_df["secondary_labels"]
secondary_labels = secondary_labels.apply(lambda x : [y.strip()[1:-1] for y in x[1:-1].split(',')])
```

```python
import glob
from tqdm.auto import tqdm
import os

path = "../input/birdclef-2021/train_short_audio/*/"
for fulldirname in tqdm(sorted(glob.glob(path))):
    label = fulldirname[41:-1]
    subpath = fulldirname+"/*.ogg"
    for idx_file, filename in enumerate(glob.glob(subpath)):
        meta_id = os.path.basename(filename)
        primary_label = label
        _secondary_labels = []
        if secondary_labels[meta_id][0] != '':
            for secondary_label in secondary_labels[meta_id]:
                _secondary_labels.append(secondary_label)

        labels = [primary_label] + _secondary_labels
        for i, v1 in enumerate(labels):
            for j, v2 in enumerate(labels):
                if i==j:
                    continue
                if v1 == v2:
                    print("error", labels)
```
