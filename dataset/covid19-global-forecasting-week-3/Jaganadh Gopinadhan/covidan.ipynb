```python
def format_date(date_string : str) -> datetime:
    """
        Format string to datetime object
        :param date_string: date string
        :returns date_obj: parsed date as datetime object
    """
    try:
        date_obj = datetime.strptime(
            date_string,
            '%d %b %Y %a'
        )
        return date_obj
    except:
        date_obj = datetime.strptime(
            date_string,
            '%Y-%m-%d'
        )
        return date_obj
```
