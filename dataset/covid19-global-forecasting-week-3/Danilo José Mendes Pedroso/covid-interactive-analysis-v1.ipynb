## COVID19 Interactive Analysis[](http://)


This jupyter notebook has the intention of helping to understand better the spreading of Corona Virus around the world.

The data used for this task is a combination of several datasets taking as reference the one avaliable in kaggle for the forecasting competition. 

Tha variables on this dataset are:

**Country**

**Date**

**Cases:** Number of cases at the time

**Deaths** Number of deaths at the time

**Cases per Day:** Number of cases in a specific date

**Deaths per Day:** Number of deaths in a specific date

**StartCases:** The subsequent days since the first case for specific country

**StartDeaths:** The subsequent days since the first death for specific country


The **StartCases** and **StartDeaths** are used to compare equal period of time between countries since first case or first death.

The data below were collected in worldbank

**Population_2018 (Millions):** The population of specific country in 2018

**Arrivals_2018 (Millions):** The number of international arrivals in a specific country in 2018

**GDP_2018 (Billions):** The GDP of a specific country in 2018

**HDI_2018:** The HDI of a specific country in 2018

**LifeExp_2017:** The life expectancy of a specific country in 2018

**UrbPop_2018 (%):** The urban population of a specific country in 2018

**Tropical:** If the country is tropical or not

**Code:** The country code is used to draw maps using plotly (I am working on it)

**Cases (Each 100000 ):** The number of cases per 100000 citizens at the time

**Mortality Rate (%):** Mortality rate based in the number of cases at the time
