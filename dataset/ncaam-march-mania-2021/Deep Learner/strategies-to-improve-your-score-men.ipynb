# Strategies:

## 1. Moderate Risk - Moderate Reward
Betting on some matches with an almost sure outcome. This does not work as well on Men's basketball as it does on women's basketball. I wouldn't recommend using it on NCAAM


## 2. Better Safe than Sorry
Always predicting the probabilities between 0.5 and 0.95. Using this strategy prevents high log loss in case of unseen / black swan event. It can work well on the upcoming 2021 tournaments.

## 3. High Risk High Reward
Using the 2 submissions to take a chance on 3 most closely matched games (~0.5 prediction by model). One submission always predicts the team with lower seed and other submission always predict the team with higher seed

## 4. Higher Risk Higher Reward (2019 3rd Place)
Instead of using the two submissions to flip one game (1 to 0), we can flip 32 most closely matched games (0.64 to 0.36). When the game is unpredictable (model prediction is around 0.5), we can pick the team with lower ID to win with 64% probability. In the second submission, we can predict that the other wins with 64% probability. Assuming that team IDs have nothing to do with winning, this is just random guessing. But if you get lucky, you can win big

## Better Safe than Sorry

```python
preds = submission.Pred.values
CLIP_AT = 0.95
safe_preds = np.where(preds > CLIP_AT, CLIP_AT, preds)
safe_preds = np.where(preds < 1 - CLIP_AT, 1 - CLIP_AT, preds)
if STRATEGY == BETTER_SAFE_THAN_SORRY: 
    submission.Pred = safe_preds
    submission.to_csv('safe_predictions.csv')
else: 
    print('I have not yet implemented the rest of the strategies on NCAAM. Check out my NCAAW notebook')
    print('https://www.kaggle.com/readoc/strategies-to-improve-your-score/')
```
