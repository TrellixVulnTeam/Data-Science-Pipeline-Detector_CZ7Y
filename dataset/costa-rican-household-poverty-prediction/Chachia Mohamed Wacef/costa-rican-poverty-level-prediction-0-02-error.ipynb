```python
from sklearn.feature_selection import SelectKBest, f_regression
```

```python
x = train.drop(['Target','Id'],axis=1)
y = train.Target
```

```python
# k is the number of features you want to select 
fs = SelectKBest(score_func=f_regression,k=100)
x_new= fs.fit_transform(x,y)
```

```python
#display the selected features
l_f = fs.get_support()
print(l_f)
```

```python
len(l_f)
```

```python
l_f2 = test.columns
```

```python
l_f2
```

```python
l_f2 = list(l_f2)
```

```python
del l_f2[len(l_f2)-1]
```

```python
#l_f2
```

```python
len(l_f2)
```

```python
fs_se = []
for i in range (len(l_f2)):
  if l_f[i] == True :
    fs_se.append(l_f2[i])
```

```python
len(fs_se)
```

```python
new_test = test[fs_se]
```

```python
fs_se.append('Target')
```

```python
new_train = train[fs_se]
```

```python
#Model Training
```

```python
#RandomForest
```

```python
x1 = new_train.drop(['Target'],axis=1)
y1 = new_train.Target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(x1,y1, test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor(n_estimators=200,criterion='mse',random_state=42,max_depth=15)
```

```python
rfr.fit(X_train,y_train)
```

```python
y_pred = rfr.predict(X_val)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_pred,y_val)
```
