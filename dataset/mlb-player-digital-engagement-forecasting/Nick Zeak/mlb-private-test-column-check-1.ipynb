```python
'''Standard imports'''
import numpy as np
import pandas as pd
import time
from itertools import chain
import json

'''Imports needed for mlb environment'''
import sys
if 'kaggle_secrets' in sys.modules:
    import mlb

'''Sorted list of columns from public test data'''
example_test_labels = [
    'awards',
    'date',
    'events',
    'games',
    'playerBoxScores',
    'playerTwitterFollowers',
    'rosters',
    'standings',
    'teamBoxScores',
    'teamTwitterFollowers',
    'transactions'
]
missing_column = [
    False,
    False,
    False,
    False
]
is_missing_values = []

column_to_test = "rosters"
'''Iterate through mlb environment'''
if 'kaggle_secrets' in sys.modules:
    # Make environment
    env = mlb.make_env()
    iter_test = env.iter_test()
    # iterate through test data
    for (test_df, sample_prediction_df) in iter_test:
        # get date column from index if date is not column
        if "date" not in list(test_df):
            test_df = test_df.reset_index()
            test_df = test_df.rename(columns={"index": "date"})
        if len(pd.DataFrame(list(chain(*[json.loads(data) for data in test_df[column_to_test].copy().dropna().to_list()])))) == 0:
            is_missing_values.append(True)
        else:
            is_missing_values.append(False)
        env.predict(sample_prediction_df)   
if np.any(np.array(is_missing_values)):
    time.sleep(600.0)
    raise ValueError
print("Complete.")
```
