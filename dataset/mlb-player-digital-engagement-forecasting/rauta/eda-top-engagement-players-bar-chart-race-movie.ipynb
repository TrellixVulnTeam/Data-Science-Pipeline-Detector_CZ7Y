```python
!pip install bar_chart_race
import bar_chart_race as bcr
import pandas as pd
import numpy as np
import json
from IPython.display import HTML
import warnings
warnings.filterwarnings('ignore')
```

```python
players = pd.read_csv('../input/mlb-player-digital-engagement-forecasting/players.csv', usecols=['playerId', 'playerName'])
train = pd.read_csv('../input/mlb-player-digital-engagement-forecasting/train.csv', usecols=['nextDayPlayerEngagement'])
for c in ['nextDayPlayerEngagement']:
    train.loc[pd.notnull(train[c]), c] = train.loc[pd.notnull(train[c]), c].map(lambda x: json.loads(x))
    
target_df = pd.DataFrame(list(np.concatenate(train['nextDayPlayerEngagement'].values)))
target_df['engagementMetricsDate'] = pd.to_datetime(target_df['engagementMetricsDate'])
target_df = target_df.merge(players, on='playerId', how='left')
target_df
```

```python
def create_bcr_html(df, target_col):
    df['cummean'] = df.groupby('playerName')[target_col].transform(lambda x: x.cumsum()) / (df.groupby('playerName').cumcount() + 1)
    df = df[df['engagementMetricsDate']>'2019-01-01']
    pivot_df = df[['engagementMetricsDate', 'playerName', 'cummean']].pivot_table(
        index='engagementMetricsDate', 
        columns='playerName',
        values='cummean',
    )
    bcr_html = bcr.bar_chart_race(df=pivot_df, filename=None, n_bars=20, figsize=(6, 6), period_length=100, title=f'{target_col}: Top 20 Engagement Player Bar Chart Race from 2019')
    return bcr_html
```

```python
bcr_html = create_bcr_html(target_df, 'target1')
HTML(bcr_html.data.replace('<video width="1008" height="864" controls autoplay loop>', '<video width="720" height="864" controls autoplay loop>'))
```
