```python
# Progress bar
from tqdm import tqdm
# ast, a module to parse string of list and dict
import ast

    

target_columns = ['nextDayPlayerEngagement']  # You can add more columns in this list but beware RAM usage # noqa E501

# target_columns = train.drop('date', axis=1).columns

for col in target_columns:
    print(f"Parse columns {col}")
    target_df = pd.DataFrame()
    target_df['date'] = train['date']
    for i in range(train.shape[0]):
        cell = train.iloc[i][col]
        if cell is not None:
            # If a cell is not None make list
            try:
                value = ast.literal_eval(cell)
            except ValueError:
                continue
            for k, v in enumerate(value):
                prefix_col = f'{col}_{k}'
                
                for key, value in v.items():
                    new_col = f'{prefix_col}_{key}'
                    if new_col not in target_df.columns:
                        train[new_col] = None
                    target_df.at[i, new_col] = value
        # Save to csv
    target_df.to_csv(f'{col}.csv')

target_df.head(5)
```
