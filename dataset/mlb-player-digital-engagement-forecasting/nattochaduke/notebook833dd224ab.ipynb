```python
train_df = pd.read_csv("../input/mlb-player-digital-engagement-forecasting/train.csv")
```

```python
first_day_df = pd.DataFrame(eval(train_df.iloc[0]['nextDayPlayerEngagement']))
player_id_top_player_t1 = first_day_df.iloc[first_day_df['target1'].idxmax()]['playerId']
player_id_random = first_day_df.iloc[801]['playerId']
```

```python
import datetime
dfs = []
for i in range(len(train_df)):
    row = train_df.iloc[i]
    date = datetime.datetime.strptime(str(row['date']), '%Y%m%d')
    day_df = pd.DataFrame(eval(row['nextDayPlayerEngagement']))
    dfs.append(day_df)
engs = pd.concat(dfs)
```

```python
engs = pd.concat(dfs)
```

```python
players = set(engs['playerId'])
averages = {}
for playerId in players:
    averages[playerId] = engs[engs['playerId'] == playerId].mean()
```

```python
def get_history(playeriD):
    history = {}
    for i in range(len(train_df)):
        row = train_df.iloc[i]
        date = datetime.datetime.strptime(str(row['date']), '%Y%m%d')
        day_df = pd.DataFrame(eval(row['nextDayPlayerEngagement']))
        top_player_row = day_df[day_df['playerId'] == playeriD]
        history[date] = [float(top_player_row[t])/day_df['target1'].sum() for t in TARGETS]
    return history
```

```python
history = get_history(player_id_random)
plt.plot(sorted(history.keys()), [history[d][0] for d in sorted(history.keys())][:100])
plt.plot(sorted(history.keys()), [history[d][1] for d in sorted(history.keys())][:100])
plt.plot(sorted(history.keys()), [history[d][2] for d in sorted(history.keys())][:100])
plt.plot(sorted(history.keys()), [history[d][3] for d in sorted(history.keys())][:100])
```
