```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.deterministic import (CalendarFourier,
                                           CalendarSeasonality,
                                           CalendarTimeTrend,
                                           DeterministicProcess)
```

```python
df_train = pd.read_csv('../input/mlb-player-digital-engagement-forecasting/train.csv', usecols=['date'])
df_train.head()
```

```python
df_train['date'] = pd.to_datetime(df_train['date'], format="%Y%m%d")
df_train.head()
```

```python
# CalendarFourier https://www.statsmodels.org/stable/generated/statsmodels.tsa.deterministic.CalendarFourier.html
#  freq"A": {"M": 12, "Q": 4},
fourier_terms = CalendarFourier(freq='A', order=4)
df_fourier = fourier_terms.in_sample(df_train['date'])
df_fourier
```

```python
df_fourier[['sin(1,freq=A-DEC)','cos(1,freq=A-DEC)']].query('date < "2019-01-01"').plot()
df_fourier[['sin(2,freq=A-DEC)','cos(2,freq=A-DEC)']].query('date < "2019-01-01"').plot()
df_fourier[['sin(3,freq=A-DEC)','cos(3,freq=A-DEC)']].query('date < "2019-01-01"').plot()
df_fourier[['sin(4,freq=A-DEC)','cos(4,freq=A-DEC)']].query('date < "2019-01-01"').plot()
```
