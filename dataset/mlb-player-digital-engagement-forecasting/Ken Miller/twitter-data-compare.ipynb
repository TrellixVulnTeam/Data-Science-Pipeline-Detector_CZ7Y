This notebook takes Mike Trouts tweets from February 2021 to the beginning of May and scales the 4 matrics (retweets, like, quotes and replies) between 0 and 100 just like the target - it then plots the different metrics and Mike Trouts targtes for the same period - I now do not think that the targets are the twitter metrics based on this.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
trout_twitter = pd.read_csv("../input/mike-trout-twitter-data/miketrout_tweets_2021.csv")
trout_twitter.head()
```

```python
engagements = pd.read_csv("../input/mlb-pdef-train-dataset/nextDayPlayerEngagement_train.csv")
```

```python
# mike trout id  = 545361
engagements = engagements[engagements['playerId'] == 545361]
engagements['year'] = pd.DatetimeIndex(engagements['engagementMetricsDate']).year
engagements['month'] = pd.DatetimeIndex(engagements['engagementMetricsDate']).month

engagements = engagements[engagements['year'] == 2021]
engagements = engagements[engagements['month'] > 1]
```

```python
#zi = (xi – min(x)) / (max(x) – min(x)) * 100

trout_twitter=trout_twitter.head(35)

trout_twitter['like_count']=(trout_twitter['like_count']-min(trout_twitter['like_count']))/(max(trout_twitter['like_count'])-min(trout_twitter['like_count']))*100
trout_twitter['retweet_count']=(trout_twitter['retweet_count']-min(trout_twitter['retweet_count']))/(max(trout_twitter['retweet_count'])-min(trout_twitter['retweet_count']))*100

trout_twitter['quote_count']=(trout_twitter['quote_count']-min(trout_twitter['quote_count']))/(max(trout_twitter['quote_count'])-min(trout_twitter['quote_count']))*100
trout_twitter['reply_count']=(trout_twitter['reply_count']-min(trout_twitter['reply_count']))/(max(trout_twitter['reply_count'])-min(trout_twitter['reply_count']))*100
```
