# Func for unpack JSON
Quoted from a [Startert Notebook](https://www.kaggle.com/ryanholbrook/getting-started-with-mlb-player-digital-engagement)

```python
# Helper function to unpack json found in daily data
def unpack_json(json_str):
    return pd.DataFrame() if pd.isna(json_str) else pd.read_json(json_str)
```

# nextDayPlayerEngagement
目的変数を含んだ重要なデータになります。<br>
You will predict traget1~4.

```python
unpack_json(training['nextDayPlayerEngagement'].iloc[0])
```
