# JSONファイルの展開用関数
JSONファイルを展開するための関数<br>
公式の[スターターノート](https://www.kaggle.com/ryanholbrook/getting-started-with-mlb-player-digital-engagement)から引用

```python
# Helper function to unpack json found in daily data
def unpack_json(json_str):
    return pd.DataFrame() if pd.isna(json_str) else pd.read_json(json_str)
```

# nextDayPlayerEngagement
目的変数を含んだ重要なデータになります。<br>
このtarget1~4を予測します

```python
nextDayPlayerEngagement = unpack_json(training['nextDayPlayerEngagement'].iloc[0])
nextDayPlayerEngagement
```
