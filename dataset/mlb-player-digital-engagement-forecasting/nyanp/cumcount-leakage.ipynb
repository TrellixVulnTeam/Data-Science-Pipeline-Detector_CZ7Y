```python
import json

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import mlb

%matplotlib inline

train = pd.read_csv('../input/mlb-player-digital-engagement-forecasting/train_updated.csv')
```

```python
train.head()
```

```python

engagements = []
for i, row in train.iterrows():
    try:
        parsed = json.loads(row['nextDayPlayerEngagement'])
        engagements.extend(parsed)
    except:
        pass

engagement_df = pd.DataFrame(engagements)
engagement_df['cumcount'] = engagement_df.groupby('engagementMetricsDate').cumcount()
sampled = engagement_df.sample(500000)
```

## Cumcount VS Target

```python
plt.scatter(sampled['cumcount'], sampled['target1'])
```
