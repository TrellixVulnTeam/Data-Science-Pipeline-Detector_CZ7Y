```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
import matplotlib.dates as mdates
pd.options.display.max_columns = 999
```

```python
data = pd.read_csv('/kaggle/input/mlb-player-digital-engagement-data-exploration/out.csv')
```

```python
data['dailyDataDate'] = pd.to_datetime(data['dailyDataDate'])
```

MLB digital engagement competition aims to predict four different engagement metrics and help to see how different factors affect fan engagement in baseball. Since the metrics are base on real life performnace and news it is hard to truly anonymize the data. Here we will look at a few selected players.

First we would look at players that are retired or no longer playing in MLB. One of example is future hall of famer Ichiro Suzuki who retired in 2019 season. Here is a plot of target metrics:

```python
player_data = data[data.playerId == 400085] #Ichiro Suzuki

fig1, f1_axes = plt.subplots(ncols=2, nrows=2, constrained_layout=True,figsize=(20, 8))

plt.suptitle('Target metric of Ichiro Suzuki')

f1_axes[0,0].plot(player_data['dailyDataDate'],player_data['target1'])
f1_axes[0,0].xaxis.set_major_locator(mdates.WeekdayLocator(interval=13))
f1_axes[0,0].xaxis.set_major_formatter(DateFormatter("%y-%m-%d"))
f1_axes[0,0].set_title('Target metric 1')


f1_axes[0,1].plot(player_data['dailyDataDate'],player_data['target2'])
f1_axes[0,1].xaxis.set_major_locator(mdates.WeekdayLocator(interval=13))
f1_axes[0,1].xaxis.set_major_formatter(DateFormatter("%y-%m-%d"))
f1_axes[0,1].set_title('Target metric 2')

f1_axes[1,0].plot(player_data['dailyDataDate'],player_data['target3'])
f1_axes[1,0].xaxis.set_major_locator(mdates.WeekdayLocator(interval=13))
f1_axes[1,0].xaxis.set_major_formatter(DateFormatter("%y-%m-%d"))
f1_axes[1,0].set_title('Target metric 3')

f1_axes[1,1].plot(player_data['dailyDataDate'],player_data['target4'])
f1_axes[1,1].xaxis.set_major_locator(mdates.WeekdayLocator(interval=13))
f1_axes[1,1].xaxis.set_major_formatter(DateFormatter("%y-%m-%d"))
f1_axes[1,1].set_title('Target metric 4')

```

From the figures only target metric 4 shows non-zero value after his retirement which may imply target metric 4 is some sort of social media engagement and independnet of players' active status, and target 1 and 3 happens sparsely while targe 2 and 4 are more continous.
