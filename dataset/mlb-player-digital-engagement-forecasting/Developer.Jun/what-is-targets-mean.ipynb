```python
# https://www.kaggle.com/chumajin/eda-of-mlb-for-starter-english-ver
# Helper function to unpack json found in daily data
def unpack_json(json_str):
    return np.nan if pd.isna(json_str) else pd.read_json(json_str)
```

**1. targets을 선수별로 나눠보기**

1. 'nextDayPlayerEngagement'의 json 형식을 분해
2.  분해한 것들을 playerId를 기준으로 다시 분해
3.  playerId가 같은 것들을 취합
4. 
 

```python
# 각 dataFrame(targets_n)의 index를 playerId로 변환하는 함수

def playerId_to_index(dataFrame):
    dataFrame.index = dataFrame['playerId']
    dataFrame.drop(['playerId'], axis = 1, inplace = True)
    return dataFrame
```

```python
# 날짜 별 targets_n 을 만들어 내는 함수
def make_targets_df(train_df, start, end):
    for i in range(start,end):
        dataFrame = unpack_json(train_df['nextDayPlayerEngagement'].iloc[i])
        dataFrame = playerId_to_index(dataFrame)
        if i == start :
            targets_df = dataFrame
        else :
            targets_df = pd.merge(targets_df, dataFrame, how='outer',on='playerId')
        if i%5 == 0:
            print(i)
    return targets_df
```
