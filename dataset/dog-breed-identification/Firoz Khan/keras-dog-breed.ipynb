```python
# Turning images into array
def prepareImages(data, m, dataset):
    
    print("Preparing images")
    
    X_train = np.zeros((m, 100, 100, 3))
    
    count = 0
    
    for fig in data['id']:
        
        #load images into images of size 100x100x3
        
        img = image.load_img("../input/dog-breed-identification/"+dataset+"/"+fig + ".jpg", target_size=(100, 100, 3))
        x = image.img_to_array(img)
        
        
        x = preprocess_input(x)
        
        X_train[count] = x
        
        if (count%500 == 0):
            print("Processing image: ", count+1, ", ", fig)
        count += 1
    
    count = 0
    
    print("Finished!")
            
    return X_train
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
# Turning labels into one hot encoder
def prepareY(Y):

    values = array(Y)
    print(values.shape)
    label_encoder = LabelEncoder()
    integer_encoded = label_encoder.fit_transform(values)
    print(integer_encoded)

    onehot_encoder = OneHotEncoder(sparse=False)
    integer_encoded = integer_encoded.reshape(len(integer_encoded), 1)
    onehot_encoded = onehot_encoder.fit_transform(integer_encoded)
    print(onehot_encoded)

    y = onehot_encoded
    print(y.shape)
    return y, label_encoder
```

```python
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Conv2D, BatchNormalization, MaxPooling2D
from keras.layers import Dropout, Activation, Dense, AveragePooling2D, Flatten
```

```python
model = Sequential()

model.add(Conv2D(32, (7,7), (1,1), input_shape = (100,100,3)))
model.add(BatchNormalization(axis = 3))
model.add(Activation("relu"))
model.add(MaxPooling2D(2,2))

model.add(Conv2D(64, (3,3), (1,1)))
model.add(Activation("relu"))
model.add(AveragePooling2D(3,3))

model.add(Flatten())
model.add(Dense(5000, activation = "relu"))
model.add(Dropout(0.6))
model.add(Dense(10222, activation = "softmax"))
```

```python
model.output_shape
```

```python
model.compile(loss = "categorical_crossentropy", optimizer = "adam", metrics = "accuracy")
```
