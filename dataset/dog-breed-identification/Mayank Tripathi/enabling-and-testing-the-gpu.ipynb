# Enabling and Testing the GPU
While working on large dataset or some critical project such as computer vision / image processing etc, we may need GPU or TPU to use along with the provided CPU.

Some how I was strugling on how to get on to it, and after reaading some articles and kaggle notebooks, I concluded the below code, which ensures that GPU is turned on and it is going to be consumed with the project we are running on it.

Note : Turning on GPU will reduce the number of CPUs available and will only speed up image processing and neural networks.

Also there is a limit on the usage of GPU, it is limited to 34 hr for a week. This quota gets reset weekly.

The limit of 30-34 hrs is based on today. In future it may or may not change.

In case you have received the message as `GPU device not found`, then dont worry. Just check if the following setting is turned on.
Follow the steps.. and re-run the code.



```python
import tensorflow as tf

device_name = tf.test.gpu_device_name()

if "GPU" not in device_name:
    print("GPU device not found")
    
print('Found GPU at: {}'.format(device_name))

print("GPU", "available (YESS!!!!)" if tf.config.list_physical_devices("GPU") else "not available :(")
```

# Step 1 : Verify if the GPU is On or not.
On the right top corner, you will have the options such as HDD; CPU; RAM and GPU as well. 
If GPU is displayed there then it is ON, and we should be good to use the GPU. Now re-run the above code, and it should find the GPU for you.
In-case it is not displayed there, then follow step # 2 to configure it.



# Step 2 : Configure the GPU.
On the same right top corner along with HDD; CPU and RAM there will be vertical 3 dots. Click on it.



After click on the 3 vertical dots, Navigate to accelerator and it will showup the window for selecting GPU / TPU.

Select the GPU, and confirm to turn-on the GPU.



Once the step 2 is done, re-run the code, and it should be using the GPU now.
