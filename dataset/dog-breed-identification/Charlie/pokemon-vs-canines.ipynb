# Introduction

Using the Keras library to detect image similarties, I will find out what the real-world canine comparisons of canine-like Pokemon are.

I'm using a pre-trained model <code>Resnet50</code> to process images of dogs and convert them into their vector form. These vectors are compared to the vector of canine-like Pokemon species and euclidean distances are computed between them. 

# Solution

```python
from keras.applications.resnet50 import ResNet50
from keras.preprocessing import image
from keras.preprocessing.image import load_img
from keras.applications.resnet50 import preprocess_input, decode_predictions
from keras.models import Model
import numpy as np
from os import listdir, walk
from os.path import isfile, join
import itertools
import sys,requests
from matplotlib import pyplot as plt
import pandas as pd
```

## Helper Functions
Here are some helper functions that I will use in this project.

```python
def getAllFilesInDirectory(directoryPath: str):
    ''' Helper function to read all the files in the directory by Kaggle '''
    return [(directoryPath + "/" + f) for f in listdir(directoryPath) if isfile(join(directoryPath, f))]

def load_image(img_path, show=False):
    ''' Helper function to load and show images using Keras'''
    img = image.load_img(img_path, target_size=(224, 224))
    img_tensor = image.img_to_array(img)
    img_tensor = np.expand_dims(img_tensor, axis=0)
    if show:
        plt.imshow(img_tensor[0]/255)                           
        plt.axis('off')
        plt.show()
```

# Output

```python
def calcSimilarity(self_vect, feature_vectors):
    ''' Calculate Euclidian distances'''
    similar: dict = {}
    keys = [k for k,v in feature_vectors.items()]
    min_dist = 10000000
    for k,v in feature_vectors.items():
       dist=np.linalg.norm(self_vect-v)
       if(dist < min_dist):
           min_dist = dist
           similar = k
    return similar 
```

```python
def predict(img_path : str, model: Model):
    img = image.load_img(img_path, target_size=(224, 224))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    return model.predict(x)
```
