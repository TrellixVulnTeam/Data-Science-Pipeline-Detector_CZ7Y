```python
df_labels = pd.read_csv('../input/dog-breed-identification/labels.csv')
df_labels.head()
```

```python
df_labels.shape
```

```python
train_file_location = '../input/dog-breed-identification/train/' 
train_data = df_labels.assign(img_path = lambda x : train_file_location + x['id'] + '.jpg')
train_data.head()
```

```python
len(train_data['breed'].unique())
```

```python
train_data['img_path'][0]
```

```python
type(train_data['img_path'][0])
```

```python
from keras.preprocessing.image import load_img,img_to_array, ImageDataGenerator
X = np.array([img_to_array(load_img(img,target_size = (128,128))) for img in train_data['img_path'].values.tolist()])
X.shape
```

```python
Y = pd.get_dummies(train_data['breed'])
Y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_val,y_train,y_val  = train_test_split(X,Y,test_size = 0.25)
```

```python
print(X_train.shape,y_train.shape)
print(X_val.shape,y_val.shape)
```

```python
from keras.utils import normalize
X_train = normalize(X_train)
X_val = normalize(X_val)
```

```python
train_datagen = ImageDataGenerator(shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)
train_datagen.fit(X_train)

train_generator = train_datagen.flow(X_train,y_train,batch_size = 32)
```

```python
from keras.models import Sequential
from keras.layers import Dense,Conv2D,MaxPool2D,Dropout,Flatten,Activation,BatchNormalization
from keras.models import model_from_json
from keras.models import load_model
```

```python
model = Sequential()

model.add(Conv2D(32,(3,3),input_shape = (128,128,3)))
model.add(Activation('elu'))

model.add(Conv2D(64,(3,3)))
model.add(Activation('elu'))

model.add(MaxPool2D(pool_size = (2,2)))

model.add(Conv2D(64,(3,3)))
model.add(Activation('elu'))

model.add(Conv2D(32,(3,3)))
model.add(Activation('elu'))

model.add(Conv2D(16,(3,3)))
model.add(Activation('elu'))

model.add(MaxPool2D(pool_size = (2,2)))

model.add(Conv2D(8,(3,3)))
model.add(Activation('elu'))

model.add(Conv2D(8,(3,3)))
model.add(Activation('elu'))

model.add(Flatten())

model.add(Dense(128))
model.add(BatchNormalization())
model.add(Activation('elu'))

model.add(Dense(100))
model.add(Activation('elu'))

model.add(Dense(100))
model.add(Activation('elu'))

model.add(Dense(100))
model.add(Activation('elu'))

model.add(Dropout(0.25))

model.add(Dense(100))
model.add(Activation('elu'))

model.add(Dense(Y.shape[1]))
model.add(Activation('softmax'))

model.summary()
```

```python
model.compile(optimizer='adam',
             loss='categorical_crossentropy',
             metrics=['accuracy'])
```

```python
tr_history = model.fit_generator(train_generator,
         validation_data=(X_val,y_val),
         epochs=50,
         verbose=2)
```

```python
loss = tr_history.history['loss']
val_loss = tr_history.history['val_loss']
epochs = range(1, len(loss) + 1)
plt.plot(epochs, loss, 'y', label='Training loss')
plt.plot(epochs, val_loss, 'r', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
acc = tr_history.history['accuracy']
val_acc = tr_history.history['val_accuracy']
epochs = range(1,len(acc)+1)
plt.plot(epochs,acc,'y',label = 'Training_acc')
plt.plot(epochs,val_acc,'r',label = 'Validation_acc')
plt.title('Training and Validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
