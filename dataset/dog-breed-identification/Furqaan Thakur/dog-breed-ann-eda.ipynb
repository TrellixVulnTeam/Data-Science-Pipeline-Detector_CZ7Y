```python
# Loading the labels file for performing EDA

import pandas as pd

labels_filepath = "../input/dog-breed-identification/labels.csv"
labels_df = pd.read_csv(labels_filepath)

labels_df.head()
```

```python
# Checking Class Balancing

print("Maximum samples for a breed: ", labels_df.breed.value_counts().max())
print("Minimum samples for a breed: ", labels_df.breed.value_counts().min())
```

```python
labels_df.breed.value_counts().plot.bar(figsize=(25,12))

'''
As it could be observed in the bar plot below, 
there is a slight imbalance in the number of samples available for each class.
It is not a significant number, the dataset looks good for further training and processing.
'''
```
