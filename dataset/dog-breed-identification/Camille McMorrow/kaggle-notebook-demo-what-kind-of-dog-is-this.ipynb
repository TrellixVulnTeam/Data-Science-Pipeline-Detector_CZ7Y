```python
import os
import pandas as pd
import numpy as np
from os.path import join
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.applications import ResNet50
from IPython.display import Image, display
from learntools.deep_learning.decode_predictions import decode_predictions

# Get list of all images in the train/ folder
image_dir = '../input/dog-breed-identification/train/'
img_paths = [join(image_dir, filename) for filename in os.listdir("../input/dog-breed-identification/train")]

# Prepare image for machine learning model
def read_and_prep_image(img_path, img_height=224, img_width=224):
    img = load_img(img_path, target_size=(img_height, img_width)) 
    output = preprocess_input(np.expand_dims(img_to_array(img), axis=0))
    return output

# Import a pre-trained model
model = ResNet50(weights='../input/resnet50/resnet50_weights_tf_dim_ordering_tf_kernels.h5')

# What does the model think the image contains?
def show_predictions(index):
    preds = model.predict(read_and_prep_image(img_paths[index]))
    most_likely_labels = decode_predictions(preds, top=3)[0]
    df = pd.DataFrame([x[1:] for x in most_likely_labels], columns=["label", "probability"])
    display(df.head())
    display(Image(img_paths[index])) 
    
print("Success! Model loaded.")
```

Alright! Now we're ready to feed the model an image. My dog's photo is number 112. Could you pass that number to the model in the next code cell and we'll see what breed he looks like?

```python
show_predictions(1)
```
