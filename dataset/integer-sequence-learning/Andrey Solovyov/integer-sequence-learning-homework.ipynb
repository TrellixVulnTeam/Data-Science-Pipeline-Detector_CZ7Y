Дан датасет из 113845 целочисленных последовательностей, предлагается любыми методами предугадать следующее число из данной последовательности
Мои варианты решения:
* Думаю, для начала все данные нужно разбить на 2 части: 
    1. В которых только положительные числа и 0, их маркер: ">=0"
    2. Остальные, маркер: "!>=0"
* Далее каждую из полученных частей еще раз разбить на 2 части:
    1. Монотонные, маркер: "m"
    2. Остальные, маркер: "!m"
* Теперь датасет разбит на 4 категории, с каждой из которых нужно работать отдельно.  
* В предложенных решениях на Kaggle использовались, в основном, следующие идеи:
    1. Поиск коэфициентов для линейной рекуррентной последовательности
        1. 1. Conclusion:
            1. 1. 1. Number of sequences in the test set: 113845
            1. 1. 1. Number of 2nd order sequences: 2121
            1. 1. 1. Number of 3rd order sequences: 1911
            1. 1. 1. Number of 4th order sequences: 1622
    2. Методом линейного дифференциирования до порядка, на котором достигается монотонность по модулю
    3. Поиском данной последовательности в библиотеке последовательностей OEIS
    4. Машинное обучение
* Мои идеи по поводу решения:
    1. Примитивная: посколько нам дана только часть последовательности, можно ее периодически продолжить и это вполне будет адекватным ответом
    2. Обобщение готовых решений:
        1. 2. Поиск коэфициентов для рекуррентной последовательности произвольного типа произвольного размера
        1. 2. Методом произвольного дифференциирования до порядка, на котором достигается монотонность по модулю, в том числе можно попробовать и скорость роста порядка дифференциирования
    3. Машинное обучение
