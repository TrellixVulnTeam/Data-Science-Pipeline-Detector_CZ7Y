````python
%%time
all_impact_list = []
for img_loc in tqdm(all_imgs, total = int(len(all_imgs))):
    gameKey = img_loc.split("/")[-1].split("_")[0]
    playID = img_loc.split("/")[-1].split("_")[1]
    view_ = img_loc.split("/")[-1].split("_")[2]
    frame_ = img_loc.split("/")[-1].split("_")[-1].replace(".png", "")
    video_ = f'{gameKey}_{playID}_{view_}.mp4'
    img = np.array(Image.open(img_loc))
    infer_ds = Dataset.from_images([img], infer_tfms)
    batch, samples = efficientdet.build_infer_batch(infer_ds)
    preds = efficientdet.predict(model=model_pr, batch=batch,  detection_threshold=0.15,
                                device = 'cuda')
    impact_dets = {}
    for lab, bbox in zip(preds[0]["labels"], preds[0]['bboxes']):
        if lab == 2:
            box_formatted = [bbox.xmin, bbox.width, bbox.ymin, bbox.height]
            inner_list = [gameKey, playID, view_, video_, frame_, int(bbox.xmin), int(bbox.width), int(bbox.ymin), int(bbox.height)]
            all_impact_list.append(inner_list)
````
