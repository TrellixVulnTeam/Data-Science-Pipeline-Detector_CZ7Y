```python
import pandas as pd
data = pd.read_csv('/kaggle/input/restaurant-revenue-prediction/train.csv.zip')
print('raw data')
print(data)
print(data.info())
```

transform Object to encoder

```python
from sklearn.preprocessing import LabelEncoder
le1=LabelEncoder()
le1.fit(data['City Group'])
data['Encoded_City Group']=le1.transform(data['City Group'])
print(data['Encoded_City Group'])

le2=LabelEncoder()
le2.fit(data['Type'])
data['Encoded_Type']=le2.transform(data['Type'])
print(data['Encoded_Type'])
```
