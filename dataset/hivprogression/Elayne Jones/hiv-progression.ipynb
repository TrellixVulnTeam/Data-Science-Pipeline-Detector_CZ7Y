```python
from sklearn.model_selection import train_test_split
X=train_data[['VL-t0', 'CD4-t0']]
Y=train_data['Resp'].values
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
knn_model = KNeighborsClassifier()
knn_model.fit(X_train, Y_train)
predicted = knn_model.predict(X_test)
print('KNN:', accuracy_score(Y_test, predicted))
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc_model = RandomForestClassifier()
rfc_model.fit(X_train, Y_train)
predicted = rfc_model.predict(X_test)
print('Random Forest:', accuracy_score(Y_test, predicted))
```

```python
svc = SVC()
svc.fit(X_train, Y_train)
predicted = svc.predict(X_test)
acc_svc = round(svc.score(X_train, Y_train) * 100, 2)
print('SVC:', accuracy_score(Y_test, predicted))
```

```python
gaussian = GaussianNB()
gaussian.fit(X_train, Y_train)
predicted = gaussian.predict(X_test)
acc_gaussian = round(gaussian.score(X_train, Y_train) * 100, 2)
print('Gaussian Naive Bayes:', accuracy_score(Y_test, predicted))
```

```python
perceptron = Perceptron()
perceptron.fit(X_train, Y_train)
predicted = perceptron.predict(X_test)
acc_perceptron = round(perceptron.score(X_train, Y_train) * 100, 2)
print('Perceptron:', accuracy_score(Y_test, predicted))
```
