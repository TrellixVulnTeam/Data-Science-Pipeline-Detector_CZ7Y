```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

import seaborn as sns 
sns.set_theme(style="darkgrid")
```

```python
TARGET = "Cover_Type"
```

```python
train = pd.read_csv("../input/tabular-playground-series-dec-2021/train.csv")
```

```python
def getLabelCount(df,target):
    return sorted([( labelValue,len(train.loc[df[target] == labelValue]) ) for labelValue in df[target].unique()])

labelCount = getLabelCount(train,TARGET)
labelCount
```

```python
ax = sns.countplot(data=train,x=TARGET)
```

```python
ax = sns.countplot(data=train,y=TARGET)
```

```python
ax = sns.countplot(x=TARGET, 
                   data=train,
                   facecolor=(0, 0, 0, 0),
                   linewidth=5,
                   edgecolor=sns.color_palette("dark", len(labelCount)))
```

```python
ax = sns.countplot(y=TARGET, 
                   data=train,
                   facecolor=(0, 0, 0, 0),
                   linewidth=5,
                   edgecolor=sns.color_palette("dark", len(labelCount)))
```
