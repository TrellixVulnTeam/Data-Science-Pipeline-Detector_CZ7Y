```python
import numpy as np
import pandas as pd
```

```python
TRAIN_PATH = "../input/tabular-playground-series-dec-2021/train.csv"
TARGET = "Cover_Type"
```

```python
train = pd.read_csv(TRAIN_PATH)
```

```python
def getLabelCount(df,target):
    return sorted([( labelValue,len(train.loc[df[target] == labelValue]) ) for labelValue in df[target].unique()])

def getLbelCount2(df,target):
    return np.unique(df[target].values, return_counts=True)
    
```

```python
labelCount = getLabelCount(train,TARGET)
labelCount
```

```python
label,labelCount = getLbelCount2(train,TARGET)
label,labelCount
```
