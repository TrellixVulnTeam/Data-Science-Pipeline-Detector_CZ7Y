```python
from IPython.display import clear_output
!pip install dabl
clear_output()

import dabl

import numpy as np
import pandas as pd
```

```python
TRAIN_PATH = "../input/tabular-playground-series-dec-2021/train.csv"
TEST_PATH = "../input/tabular-playground-series-dec-2021/test.csv"
TARGET = "Cover_Type"
SAMPLE_SUBMISSION_PATH = "../input/tabular-playground-series-dec-2021/sample_submission.csv"
SUBMISSION_PATH = "submission.csv"
```

```python
train = pd.read_csv(TRAIN_PATH)
test = pd.read_csv(TEST_PATH)
train.head()
```

```python
dabl_train = dabl.clean(train, verbose=1)
```

```python
dabl_test = dabl.clean(test, verbose=1)
```

```python
dabl_train.columns
```

# check column type 

```python
types = dabl.detect_types(dabl_train)
types
```

# dabl basic plot

```python
dabl.plot(train, target_col=TARGET)
```

# build base best model

```python
clf = dabl.SimpleClassifier(random_state=42).fit(train, target_col=TARGET) 
```

# evaluation model

```python
clf.current_best_
```

# get best model 

```python
clf.est_
```

# predict test data using best model

```python
pred_test = clf.est_.predict(test)
```

# submission

```python
sub = pd.read_csv(SAMPLE_SUBMISSION_PATH)
sub[TARGET] = (pred_test- 0.5).astype(int) + 1
sub.to_csv(SUBMISSION_PATH,index=False)
sub.head()
```
