# Latent Dirichlet Allocation

```python
import numpy as np
import pandas as pd 
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
TARGET = "Cover_Type"
# RANDOM_SEED = 2021
# SAMPLE_NUM = 1000
```

```python
train = pd.read_csv("../input/tabular-playground-series-dec-2021/train.csv")
```

```python
labels = np.unique(train[TARGET].values)
labelcount = len(labels)
N_COMPONENT = labelcount - 1

lda_data = LDA(n_components=N_COMPONENT).fit_transform(train.drop(columns=TARGET),train[TARGET])
```

```python
lda_data[:5]
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x = lda_data[:, 0], y = lda_data[:, 1], hue = TARGET, data=train)
```
