A very simple elevation model: Spruce/Fir (`Cover_Type`=1) above 10,000 ft (3048m) and Lodgepole Pine (`Cover_Type`=2) below 10,000 ft.

```python
import numpy  as np
import pandas as pd

test = pd.read_csv("../input/tabular-playground-series-dec-2021/test.csv")
test["Cover_Type"] = np.where(test['Elevation'] > 3048,1,2)
test[["Id","Cover_Type"]].to_csv('submission.csv',index=False)
```
