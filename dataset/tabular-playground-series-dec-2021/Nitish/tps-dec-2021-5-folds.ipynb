### This code is to create 5 folds of data

*Thanks to [Abhishek](https://www.kaggle.com/abhishek) and his [notebook](https://www.kaggle.com/abhishek/step-1-create-folds)*

```python
# Import Required libraries
import numpy as np
import pandas as pd

from sklearn import datasets
from sklearn import model_selection
```

```python
# Create a function to split data into multiple folds
# Since our target variable is multiclass , we can use each class as bins

def create_folds(data, num_splits):
    # we create a new column called kfold and fill it with -1
    data["kfold"] = -1
    
    # the next step is to randomize the rows of the data
    data = data.sample(frac=0.5,random_state = np.random.RandomState()).reset_index(drop=True)

    # initiate the kfold class from model_selection module
    kf = model_selection.StratifiedKFold(n_splits=num_splits)
    
    # fill the new kfold column
    # note that, instead of targets, we use bins!
    for f, (t_, v_) in enumerate(kf.split(X=data, y=data.Cover_Type.values)):
        data.loc[v_, 'kfold'] = f

    # return dataframe with folds
    return data
```

```python
# read training data
df = pd.read_csv("../input/tabular-playground-series-dec-2021/train.csv")

# create folds
df = create_folds(df, num_splits=20)
```
