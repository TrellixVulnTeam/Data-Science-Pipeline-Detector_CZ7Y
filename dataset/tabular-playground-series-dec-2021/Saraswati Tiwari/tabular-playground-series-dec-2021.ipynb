```python
import datatable as dt
import pandas as pd
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import LabelEncoder
```

```python
df = dt.fread('../input/tabular-playground-series-dec-2021/train.csv').to_pandas()
```

```python
le = LabelEncoder()
```

```python
for i in range(1,5):
    df[f'Wilderness_Area{i}'] = le.fit_transform(df[f'Wilderness_Area{i}'])
for i in range(1,41):
    df[f'Soil_Type{i}'] = le.fit_transform(df[f'Soil_Type{i}'])
    
df['Cover_Type'] = le.fit_transform(df['Cover_Type'])
```

```python
df.shape
```

```python
df.columns[-1]
```

```python
def create_folds(data, num_splits):
    data["kfold"] = -1
    kf = StratifiedKFold(n_splits=num_splits, shuffle=True, random_state=11)
    for f, (t_, v_) in enumerate(kf.split(X=data, y=data['Cover_Type'])):
        data.loc[v_, 'kfold'] = f
    return data
```

```python
df_5 = create_folds(df.copy(), num_splits=5)
df_10 = create_folds(df.copy(), num_splits=10)
```
