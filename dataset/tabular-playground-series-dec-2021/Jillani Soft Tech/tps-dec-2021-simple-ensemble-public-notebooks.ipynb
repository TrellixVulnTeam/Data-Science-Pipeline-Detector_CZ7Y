# Import Some Main Libs

```python
df_1 = pd.read_csv("../input/tps-12-nn-tpu-pseudolabeling-0-95690/tps12-pseudeo-submission.csv")
df_2 = pd.read_csv("../input/tps-12-nn-tpu-pseudolabeling-0-95690/tps12-pseudeo-submission.csv")
df_3 = pd.read_csv("../input/tps-12-g-res-variable-selection-nn-keras/baseline_nn.csv")
df_4 = pd.read_csv("../input/tps202112-reasonable-xgboost-model/submission.csv")
df_5 = pd.read_csv("../input/tps202112-reasonable-xgboost-model/submission.csv")
df_6 = pd.read_csv("../input/tps202112-reasonable-xgboost-model/submission.csv")
df_7 = pd.read_csv("../input/tps-12-super-fast-blending-tool/tps12-fast-blend.csv")

submission = pd.read_csv('../input/tabular-playground-series-dec-2021/sample_submission.csv')

predictions = [df_1, df_2, df_3, df_4, df_5, df_6, df_7]
```

```python
results = pd.DataFrame()
for i, ds in enumerate(predictions):
    results[f'p{i+1}'] = ds['Cover_Type']

print(results.shape)
results.head()
```

```python
%%time
results["ensemble"] = stats.mode(np.array(results), axis=1)[0]
results.head()
```

```python
def nunique(a, axis):
    return (np.diff(np.sort(a,axis=axis),axis=axis)!=0).sum(axis=axis)+1
```

```python
results["dif"] = nunique(results.iloc[:,:len(predictions)].values,1) - 1
results.head()
```

```python
results.dif.value_counts()
```

# Now Submission

```python
submission['Cover_Type'] = results["ensemble"]
submission.to_csv("submission.csv", index=False)
submission.head()
```

```python
plt.figure(figsize=(10,5))
ax = sns.countplot(x=submission.Cover_Type)
plt.title("Predictions")
plt.xlabel("Cover Type")
ax.bar_label(ax.containers[0])
plt.show()
```
