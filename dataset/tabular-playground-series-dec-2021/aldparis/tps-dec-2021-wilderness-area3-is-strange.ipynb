Can we trust `Wilderness_Area3`?

```python
import numpy as np
import pandas as pd

pd.set_option('max_columns', 100)
pd.set_option('max_rows', 200)

import datatable as dt

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = dt.fread("../input/tabular-playground-series-dec-2021/train.csv").to_pandas().set_index("Id")
test = dt.fread("../input/tabular-playground-series-dec-2021/test.csv").to_pandas().set_index("Id")
```

`Cover_Type` and `Wilderness_Area3` are a little bit correlated : there are less `Cover_type` *1* when `Wilderness_Area3` is *True* than when it is *False*.<br>
And there are more `Wilderness_Area3` *False* in **test** than in **train**.<br>
`Wilderness_Area3` and `Wilderness_Area1` are also correlated in **train** but there are not in **test**. In **train** only :
* `Wilderness_Area3` = True implies that `Wilderness_Area1` = False, 
* `Wilderness_Area1` = True implies that `Wilderness_Area3` = False.

`Cover_Type` and `Wilderness_Area1` are a little bit correlated too : there are no `Cover_type` *3* when `Wilderness_Area1` is *True*.<br>

```python
f = "Wilderness_Area3"
fig, ax = plt.subplots(nrows = 5, ncols = 2, figsize=(17, 5*5))
plt.subplots_adjust(hspace = 0.7, wspace = 0.1)

sns.countplot(x='Cover_Type', data=train.loc[train[f]==0]
        , order=sorted(train['Cover_Type'].unique()), orient = "h", ax=ax[0, 0], color="blue")
sns.countplot(x='Cover_Type', data=train.loc[train[f]==1]
        , order=sorted(train['Cover_Type'].unique()), orient = "h", ax=ax[0, 1], color="blue")

ax[0, 0].set_title(f"Cover_Type with {f}==0 in train", weight='bold')
ax[0, 1].set_title(f"Cover_Type with {f}==1 in train", weight='bold')

sns.countplot(x=f, data=train, order=sorted(train[f].unique()), orient = "h", ax=ax[1, 0], color="blue")
sns.countplot(x=f, data=test, order=sorted(test[f].unique()), orient = "h", ax=ax[1, 1], color="green")
ax[1, 1].set_title(f"{f} in test", weight='bold')
ax[1, 0].set_title(f"{f} in train", weight='bold')

sns.countplot(x='Wilderness_Area1', data=train.loc[train[f]==0]
        , order=sorted(train['Wilderness_Area1'].unique()), orient = "h", ax=ax[2, 0], color="blue")
sns.countplot(x='Wilderness_Area1', data=train.loc[train[f]==1]
        , order=sorted(train['Wilderness_Area1'].unique()), orient = "h", ax=ax[2, 1], color="blue")

ax[2, 0].set_title(f"Wilderness_Area1 with {f}==0 in train", weight='bold')
ax[2, 1].set_title(f"Wilderness_Area1 with {f}==1 in train", weight='bold')

sns.countplot(x='Wilderness_Area1', data=test.loc[test[f]==0]
        , order=sorted(test['Wilderness_Area1'].unique()), orient = "h", ax=ax[3, 0], color="green")
sns.countplot(x='Wilderness_Area1', data=test.loc[test[f]==1]
        , order=sorted(test['Wilderness_Area1'].unique()), orient = "h", ax=ax[3, 1], color="green")

ax[3, 0].set_title(f"Wilderness_Area1 with {f}==0 in test", weight='bold')
ax[3, 1].set_title(f"Wilderness_Area1 with {f}==1 in test", weight='bold')


f = "Wilderness_Area1"
sns.countplot(x='Cover_Type', data=train.loc[train[f]==0]
        , order=sorted(train['Cover_Type'].unique()), orient = "h", ax=ax[4, 0], color="blue")
sns.countplot(x='Cover_Type', data=train.loc[train[f]==1]
        , order=sorted(train['Cover_Type'].unique()), orient = "h", ax=ax[4, 1], color="blue")

ax[4, 0].set_title(f"Cover_Type with {f}==0 in train", weight='bold')
ax[4, 1].set_title(f"Cover_Type with {f}==1 in train", weight='bold');
```
