```python
trainset = train[['Wilderness_Area4','Soil_Type39','Wilderness_Area3','Soil_Type10','Soil_Type40','Cover_Type']]
print(trainset)
```

```python
X = trainset.drop(columns='Cover_Type', axis=1)
Y = trainset['Cover_Type']

print(X)
print(Y)
```

```python
print(X.shape)
```

**BernoulliNB**

```python
from sklearn.naive_bayes import BernoulliNB
model_nb=BernoulliNB()
model_nb.fit(X,Y)
```

**Decision Tree**

```python
from sklearn.tree import DecisionTreeClassifier
model_dt=DecisionTreeClassifier(criterion='entropy')
model_dt.fit(X,Y)
```

Test dataset

```python
test = pd.read_csv('../input/tabular-playground-series-dec-2021/test.csv')
```

```python
test.head()
```

```python
test.shape
```

```python
testset = test[['Wilderness_Area4','Soil_Type39','Wilderness_Area3','Soil_Type10','Soil_Type40']]
print(testset)
```

```python
testset.isna().sum()
```

```python
testset.info()
```

NB submission

```python
pred = model_nb.predict(testset)
```

```python
sub = pd.read_csv('../input/tabular-playground-series-dec-2021/sample_submission.csv')
```

```python
sub['Cover_Type']=pred
sub.to_csv('BernoulliNB.csv', index=False)
```
