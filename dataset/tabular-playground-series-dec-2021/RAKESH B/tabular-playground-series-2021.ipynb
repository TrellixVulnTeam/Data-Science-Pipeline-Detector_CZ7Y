```python
col_soil = [col for col in train_data.columns if "Soil" not in col]
col_soil
```

```python
col_soil_test = [col for col in test_data.columns if "Soil" not in col]
col_soil_test
```

```python
data_train = train_data[col_soil]
data_train.head()
```

```python
data_test = test_data[col_soil_test]
data_test.head()
```

```python
X = data_train.drop("Cover_Type",axis=1)
y = data_train["Cover_Type"]
```

```python
y.unique()
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X,y)
```

```python
pred_val = list(lr.predict(data_test))
```

```python
pred_val[:10]
```

```python
scores = []
for i in pred_val:
    if 0 < i < 1.5:
        scores.append(1)
    elif 1.5 < i < 2.5:
        scores.append(2)
    elif 2.5 < i < 3.5:
        scores.append(3)
    elif 3.5 < i < 4.5:
        scores.append(4)
    elif 4.5 < i < 5.5:
        scores.append(5)
    elif 5.5 < i < 6.5:
        scores.append(6)
    else:
        scores.append(7)
scores[:10]
```

```python
data_test["Cover_Type"] = scores
total_data = data_test[["Id","Cover_Type"]]
total_data.head()
```

```python
total_data.to_csv('sample_submission.csv',index=False)
```
