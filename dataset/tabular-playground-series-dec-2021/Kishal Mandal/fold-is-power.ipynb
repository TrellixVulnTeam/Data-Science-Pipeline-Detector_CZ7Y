```python
import datatable as dt
import pandas as pd
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import LabelEncoder
```

```python
df = dt.fread('../input/tabular-playground-series-dec-2021/train.csv').to_pandas()
```

```python
df.head()
```

```python
def create_folds(data, num_splits):
    data["kfold"] = -1
    kf = StratifiedKFold(n_splits=num_splits, shuffle=True, random_state=11)
    for f, (t_, v_) in enumerate(kf.split(X=data, y=data['Cover_Type'])):
        data.loc[v_, 'kfold'] = f
    return data
```

```python
df_5 = create_folds(df.copy(), num_splits=5)
df_10 = create_folds(df.copy(), num_splits=10)
```

```python
df_5['kfold'].value_counts()
```

```python
df_5.to_csv('5fold.csv', index=False)
df_10.to_csv('10fold.csv', index=False)
```
