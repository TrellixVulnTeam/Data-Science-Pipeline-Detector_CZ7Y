```python
#from this we come to know that there are two columns that can be excluded as they have only one value : soil_type7  and soil_type15
final = data.drop(['Soil_Type15','Soil_Type7'],axis=1)
```

```python
final.columns
```

```python
train_yy = final.Cover_Type
train_yy.count()
```

```python
train_Xx= final.copy()
train_X1=train_Xx.drop('Cover_Type',axis =1)
```

```python
y= train_yy
X=train_X1
```

```python
y.count()
```

```python
X.count()
```

```python
from sklearn.model_selection import train_test_split

train_X,val_X,train_y,val_y = train_test_split(X,y,train_size =0.85,test_size =0.15,random_state=1)
```

```python
# now let us read the test data 
test_X1 = pd.read_csv('/kaggle/input/tabular-playground-series-dec-2021/test.csv')
```

```python
test_X1.columns
```

```python
final_test = test_X1.drop(['Soil_Type15','Soil_Type7'],axis=1)
```

```python
final_test.head()
```

```python
final_test.columns
```

```python
from xgboost import XGBClassifier
model2 = XGBClassifier(n_estimators=400,learning_rate =0.7,max_depth=7,colsample_bytree=0.7,gamma =1)
model2.fit(train_X,train_y,early_stopping_rounds = 5,eval_set=[(val_X,val_y)],verbose=True)
```

```python
model2.predict(val_X)

#OUTPUT : -  array([2, 2, 1, ..., 2, 1, 3], dtype=int64)
```

```python
model2.score(val_X,val_y)
#OUTPUT : - 0.9610533333333333
```

```python
pred2=model2.predict(final_test)
```

```python
out= pd.DataFrame({'Id':final_test.Id,'Cover_Type':pred2})
```

```python
out.to_csv('submission_final.csv',index=False)
```
