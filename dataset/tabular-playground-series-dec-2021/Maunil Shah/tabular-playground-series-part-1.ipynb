# PART - I CREATING A PROCESSED TRAIN.CSV

```python
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
train_df = pd.read_csv('../input/tabular-playground-series-dec-2021/train.csv')
train_df.head()
```

```python
train_df.shape
```

```python
# Converts multiple Soil types column to one column
# Also there are 40 types of soil
def data_squezer(row):
    a = None
    for i in range(1,41):
        a = 'Soil_Type'+str(i)
        if row[a] == 1:
            break
    return i
```

```python
train_df['soil_type'] = [data_squezer(train_df.iloc[i]) for i in tqdm(range(len(train_df)))]
```

```python
train_df.head()
```

```python
del train_df['Id']
```

```python
a = None
for i in range(1,41):
    a = f'Soil_Type{i}'
    del train_df[a]
```

```python
train_df.head()
```

```python
def data_squezer_2(row):
    a = None
    for i in range(1,5):
        a = 'Wilderness_Area'+str(i)
        if row[a] == 1:
            break
    return i
```

```python
train_df['wilderness_area'] = [data_squezer_2(train_df.iloc[i]) for i in tqdm(range(len(train_df)))]
```

```python
a = None
for i in range(1,5):
    a = f'Wilderness_Area{i}'
    del train_df[a]
```
