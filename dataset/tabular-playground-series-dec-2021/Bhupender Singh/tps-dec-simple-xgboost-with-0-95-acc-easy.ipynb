<p style="background-color:#C9DFEC;font-family:newtimeroman;color:#033E3E;font-size:200%;text-align:center;border-radius:40px 40px;">IMPORTING LIBRARIES</p>

```python
import numpy as np 
import pandas as pd

from warnings import filterwarnings
filterwarnings('ignore')
```

<p style="background-color:#C9DFEC;font-family:newtimeroman;color:#033E3E;font-size:200%;text-align:center;border-radius:40px 40px;">REDUCING MEMORY USAGE</p>

```python
def reduce_memory_usage(df):
    
    for col in df.columns:
        col_type = df[col].dtype
        
        if col_type != 'object':
            c_min = df[col].min()
            c_max = df[col].max()
            
            if str(col_type)[:3] == 'int':
                if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                    df[col] = df[col].astype(np.int8)
                elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                    df[col] = df[col].astype(np.int16)
                elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
                    df[col] = df[col].astype(np.int32)
                elif c_min > np.iinfo(np.int64).min and c_max < np.iinfo(np.int64).max:
                    df[col] = df[col].astype(np.int64)
            
            else:
                if c_min > np.finfo(np.float16).min and c_max < np.finfo(np.float16).max:
                    df[col] = df[col].astype(np.float16)
                elif c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max:
                    df[col] = df[col].astype(np.float32)
                else:
                    pass
        else:
            df[col] = df[col].astype('category')
    
    return df
```

<p style="background-color:#C9DFEC;font-family:newtimeroman;color:#033E3E;font-size:200%;text-align:center;border-radius:40px 40px;">READING THE DATA</p>

```python
train = pd.read_csv('../input/tabular-playground-series-dec-2021/train.csv')
reduce_memory_usage(train)
test = pd.read_csv('../input/tabular-playground-series-dec-2021/test.csv')
reduce_memory_usage(test)
submission = pd.read_csv('../input/tabular-playground-series-dec-2021/sample_submission.csv')
```

<p style="background-color:#C9DFEC;font-family:newtimeroman;color:#033E3E;font-size:200%;text-align:center;border-radius:40px 40px;">START PREPROCESSING</p>

```python
x = train.drop(columns=['Id','Cover_Type'])
y = train['Cover_Type']
test_df = test.drop(columns=['Id'])
```

<p style="background-color:#C9DFEC;font-family:newtimeroman;color:#033E3E;font-size:200%;text-align:center;border-radius:40px 40px;">SPLITING THE DATASETS</p>

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(x, y, test_size=0.2, random_state=123, shuffle =True)
```

<p style="background-color:#C9DFEC;font-family:newtimeroman;color:#033E3E;font-size:200%;text-align:center;border-radius:40px 40px;">APPLYING XGBOOST CLASSIFIER</p>

```python
from xgboost import XGBClassifier
model = XGBClassifier(learning_rate=0.3,tree_method='gpu_hist',random_state=0)
model.fit(X_train,y_train)
```

<p style="background-color:#C9DFEC;font-family:newtimeroman;color:#033E3E;font-size:200%;text-align:center;border-radius:40px 40px;">CHECKING ACCURACY</p>

```python
y_pred=model.predict(X_val)

from sklearn.metrics import accuracy_score
print('Accuracy Score : ',accuracy_score(y_val, y_pred))
```

<p style="background-color:#C9DFEC;font-family:newtimeroman;color:#033E3E;font-size:200%;text-align:center;border-radius:40px 40px;">PREDICTING TEST DATASET</p>

```python
y_pred = model.predict(test_df)
```

<p style="background-color:#C9DFEC;font-family:newtimeroman;color:#033E3E;font-size:200%;text-align:center;border-radius:40px 40px;">SUBMISSION</p>

```python
submission['Cover_Type'] = y_pred
submission.to_csv("submission.csv",index=False)
submission.head()
```
