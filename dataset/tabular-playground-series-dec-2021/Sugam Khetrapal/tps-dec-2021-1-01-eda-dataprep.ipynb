# Import Libraries

```python
from dataprep.eda import plot, plot_correlation, create_report, plot_missing
from pandas import read_csv
```

# Load Dataset

```python
# Load dataset
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
        
filename_training          = '/kaggle/input/tabular-playground-series-dec-2021/train.csv'
dataset_training           = read_csv(filename_training)
dataset_training_sample    = dataset_training.sample(frac=0.50)
```

# Print Report

```python
plot(dataset_training_sample)
```
