# Problem Statement

In this notebook we will explore how to do Automatic Exploratory Data Analysis (EDA) on the Tabular Playground Series December 2021 using [pandas profiling](https://github.com/pandas-profiling/pandas-profiling).

We are going to cover the following steps:
1. Import Libraries
2. Load dataset
3. Print Report
4. References

Let's get started.

# Import Libraries

```python
from pandas_profiling import ProfileReport
from pandas import read_csv
```

# Load dataset

```python
# Load dataset
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
        
filename_training          = '/kaggle/input/tabular-playground-series-dec-2021/train.csv'
dataset_training           = read_csv(filename_training)
dataset_training_sample    = dataset_training.sample(frac=0.25)
```

# Print Report

```python
report = ProfileReport(dataset_training_sample)
```
