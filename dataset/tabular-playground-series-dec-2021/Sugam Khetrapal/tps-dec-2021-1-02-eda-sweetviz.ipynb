# Import Libraries

```python
from pandas import read_csv
import sweetviz as sv
```

# Load Dataset

```python
# Load dataset
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
        
filename_training          = '/kaggle/input/tabular-playground-series-dec-2021/train.csv'
dataset_training           = read_csv(filename_training)
dataset_training_sample    = dataset_training.sample(frac=0.10)
```

# Print Report

```python
advert_report = sv.analyze([dataset_training_sample, 'Cover_Type'])
```

```python
advert_report.show_html()
```
