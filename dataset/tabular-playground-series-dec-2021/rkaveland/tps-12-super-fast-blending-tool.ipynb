```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings("ignore")
```

```python
sub1 = '../input/tps12-pseudolabels/subs/subs/0.95690.csv'
sub2 = '../input/tps12-pseudolabels/subs/subs/0.95687.csv'
sub3 = '../input/tps12-pseudolabels/subs/subs/0.95686.csv'
sub4 = '../input/tps12-pseudolabels/subs/subs/0.95683.csv'
sub5 = '../input/tps12-pseudolabels/subs/subs/0.95682.csv'
sub6 = '../input/tps12-pseudolabels/subs/subs/0.95680.csv'
sub7 = '../input/tps12-pseudolabels/subs/subs/0.95679.csv'
sub8 = '../input/tps12-pseudolabels/subs/subs/0.95675.csv'
sub9 = '../input/tps12-pseudolabels/subs/subs/0.95674.csv'
sub10 = '../input/tps12-pseudolabels/subs/subs/0.95673.csv'
sub11 = '../input/tps12-pseudolabels/subs/subs/0.95671.csv'


subs = [pd.read_csv(f).Cover_Type.to_numpy() for f in [sub1, sub2, sub3, sub4, sub5, sub6, sub7, sub8, sub9, sub10, sub11]]
```

```python
%%time

# One counter per class. tally[i, j] is the count of votes for class j for sample i
tally = np.zeros((len(subs[0]), 8))
ix = np.arange(tally.shape[0])
for sub in subs:
    tally[ix, sub] += 1
# get the j that maximizes tally[i, j] for each i
blend = tally.argmax(axis=1)
```

```python
sub = pd.read_csv('../input/tabular-playground-series-dec-2021/sample_submission.csv').assign(
    Cover_Type=blend
)
sub.to_csv("tps12-fast-blend.csv", index=False)
sub.tail(3)
```
