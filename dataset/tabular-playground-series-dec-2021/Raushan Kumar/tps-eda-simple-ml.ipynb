```python
independent_variable = train_forest.drop('Cover_Type', axis = 1)
dependent_variable = train_forest['Cover_Type']
```

```python
independent_variable.columns
```

```python
X = independent_variable.copy()
y = dependent_variable.copy()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state = 0)
```

```python
print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)
```

```python
# from sklearn.linear_model import LinearRegression
# reg = LinearRegression()
# reg.fit(X_train, y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
reg_r = RandomForestClassifier()
reg_r.fit(X_train, y_train)
```

```python
y_predict = reg_r.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error,accuracy_score
```

```python
r2_score(y_test, y_predict)
```
