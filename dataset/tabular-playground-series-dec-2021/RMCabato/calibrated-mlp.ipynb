```python
import numpy as np
import pandas as pd 
from numpy import mean
from numpy import std

from sklearn.model_selection import train_test_split
from collections import Counter
from sklearn.utils import shuffle

from sklearn.metrics import accuracy_score

import time
import gc

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
train = pd.read_csv('../input/tabular-playground-series-dec-2021/train.csv')
```

```python
train.drop(columns=['Id','Soil_Type7', 'Soil_Type15'], inplace=True)
```

```python
one_train_df    =shuffle( train.loc[train['Cover_Type'] == 1], random_state=2520)
two_train_df    =shuffle( train.loc[train['Cover_Type'] == 2], random_state=2520)
three_train_df  =shuffle( train.loc[train['Cover_Type'] == 3], random_state=2520)
four_train_df   =shuffle( train.loc[train['Cover_Type'] == 4], random_state=2520)
six_train_df    =shuffle( train.loc[train['Cover_Type'] == 6], random_state=2520)
seven_train_df  =shuffle( train.loc[train['Cover_Type'] == 7], random_state=2520)
```

```python
one_train_df   = one_train_df.append([train.loc[train['Cover_Type'] == 5]]*20, ignore_index=True)
two_train_df   = two_train_df.append([train.loc[train['Cover_Type'] == 5]]*20, ignore_index=True)
three_train_df = three_train_df.append([train.loc[train['Cover_Type'] == 5]]*20, ignore_index=True)
four_train_df  = four_train_df.append([train.loc[train['Cover_Type'] == 5]]*20, ignore_index=True)
six_train_df   = six_train_df.append([train.loc[train['Cover_Type'] == 5]]*20, ignore_index=True)
seven_train_df = seven_train_df.append([train.loc[train['Cover_Type'] == 5]]*20, ignore_index=True)
```

```python
one_train_df.shape, two_train_df.shape, three_train_df.shape, four_train_df.shape, six_train_df.shape, seven_train_df.shape
```

```python
Yy_one_train_df    =one_train_df.pop('Cover_Type')
Yy_two_train_df    =two_train_df.pop('Cover_Type')
Yy_three_train_df  =three_train_df.pop('Cover_Type')
Yy_four_train_df   =four_train_df.pop('Cover_Type')
Yy_six_train_df    =six_train_df.pop('Cover_Type')
Yy_seven_train_df  =seven_train_df.pop('Cover_Type')
```

```python
Yy_one_train_df.shape, Yy_two_train_df.shape, Yy_three_train_df.shape, Yy_four_train_df.shape, Yy_six_train_df.shape, Yy_seven_train_df.shape
```

```python
Xx_one_train_df   =one_train_df
Xx_two_train_df   =two_train_df
Xx_three_train_df =three_train_df
Xx_four_train_df  =four_train_df
Xx_six_train_df   =six_train_df
Xx_seven_train_df =seven_train_df
```

```python
Xx_one_train_df.shape, Xx_two_train_df.shape, Xx_three_train_df.shape, Xx_four_train_df.shape, Xx_six_train_df.shape, Xx_seven_train_df.shape
```

```python
X_train_one, X_valid_one, y_train_one, y_valid_one = train_test_split(Xx_one_train_df,Yy_one_train_df,test_size=0.2520, random_state= 2520)
X_train_two, X_valid_two, y_train_two, y_valid_two = train_test_split(Xx_two_train_df,Yy_two_train_df,test_size=0.2520, random_state= 2520)
X_train_three, X_valid_three, y_train_three, y_valid_three = train_test_split(Xx_three_train_df,Yy_three_train_df,test_size=0.2520, random_state= 2520)
X_train_four, X_valid_four, y_train_four, y_valid_four = train_test_split(Xx_four_train_df,Yy_four_train_df,test_size=0.2520, random_state= 2520)
X_train_six, X_valid_six, y_train_six, y_valid_six = train_test_split(Xx_six_train_df,Yy_six_train_df,test_size=0.2520, random_state= 2520)
X_train_seven, X_valid_seven, y_train_seven, y_valid_seven = train_test_split(Xx_seven_train_df,Yy_seven_train_df,test_size=0.2520, random_state= 2520)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler() 
X_train_one = scaler.fit_transform(X_train_one) #  transform
X_valid_one = scaler.transform(X_valid_one) #  transform

X_train_two = scaler.fit_transform(X_train_two) #  transform
X_valid_two = scaler.transform(X_valid_two) #  transform

X_train_three = scaler.fit_transform(X_train_three) #  transform
X_valid_three = scaler.transform(X_valid_three) #  transform

X_train_four = scaler.fit_transform(X_train_four) #  transform
X_valid_four = scaler.transform(X_valid_four) #  transform

X_train_six = scaler.fit_transform(X_train_six) #  transform
X_valid_six = scaler.transform(X_valid_six) #  transform

X_train_seven = scaler.fit_transform(X_train_seven) #  transform
X_valid_seven = scaler.transform(X_valid_seven) #  transform
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.calibration import CalibratedClassifierCV
from sklearn.neural_network import MLPClassifier 
```

```python
model_one_lr=LogisticRegression()
model_two_lr=LogisticRegression()
model_three_lr=LogisticRegression()
model_four_lr=LogisticRegression()
model_six_lr=LogisticRegression()
model_seven_lr=LogisticRegression()
```

```python
fit_model_one_lr   = model_one_lr.fit(X_train_one, y_train_one)
fit_model_two_lr   = model_two_lr.fit(X_train_two, y_train_two)
fit_model_three_lr = model_three_lr.fit(X_train_three, y_train_three)
fit_model_four_lr  = model_four_lr.fit(X_train_four, y_train_four)
fit_model_six_lr   = model_six_lr.fit(X_train_six, y_train_six)
fit_model_seven_lr = model_seven_lr.fit(X_train_seven, y_train_seven)
```

```python
pred_one  =fit_model_one_lr.predict(X_valid_one)
pred_two  =fit_model_two_lr.predict(X_valid_two)
pred_three=fit_model_three_lr.predict(X_valid_three)
pred_four =fit_model_four_lr.predict(X_valid_four)
pred_six  =fit_model_six_lr.predict(X_valid_six)
pred_seven=fit_model_seven_lr.predict(X_valid_seven)
```

```python
model_one_acc   = accuracy_score(y_valid_one, pred_one)
model_two_acc   = accuracy_score(y_valid_two, pred_two)
model_three_acc = accuracy_score(y_valid_three, pred_three)
model_four_acc  = accuracy_score(y_valid_four, pred_four)
model_six_acc   = accuracy_score(y_valid_six, pred_six)
model_seven_acc = accuracy_score(y_valid_seven, pred_seven)
```

```python
print(model_one_acc, model_two_acc, model_three_acc, model_four_acc, model_six_acc, model_seven_acc)
```

```python
model_coe=np.concatenate((fit_model_one_lr.coef_,
                          fit_model_two_lr.coef_,
                          fit_model_three_lr.coef_,
                          fit_model_four_lr.coef_,
                          fit_model_six_lr.coef_,
                          fit_model_seven_lr.coef_),axis=0)
```

```python
model_coe.shape
```

```python
model_inter=np.concatenate((fit_model_one_lr.intercept_,
                           fit_model_two_lr.intercept_,
                           fit_model_three_lr.intercept_,
                           fit_model_four_lr.intercept_,
                           fit_model_six_lr.intercept_,
                           fit_model_seven_lr.intercept_))
```

```python
model_inter.shape
```

```python
model_inter
```

```python
model_cla=np.concatenate((fit_model_one_lr.classes_,
                          fit_model_two_lr.classes_,
                          fit_model_three_lr.classes_,
                          fit_model_four_lr.classes_,
                          fit_model_six_lr.classes_,
                          fit_model_seven_lr.classes_))
```

```python
model_cla
```

```python
model_Aa = MLPClassifier(max_iter=50000)
```

```python
model_Aa.intercept_=model_inter 
```

```python
model_Aa.coef_=model_coe
```

```python
model_Aa.classes_=model_cla
```

```python
cccv_final =CalibratedClassifierCV(base_estimator=model_Aa,cv=10, method='sigmoid')
```

```python
X_train_final=np.concatenate((X_train_one, 
                            X_train_two, 
                            X_train_three, 
                            X_train_four, 
                            X_train_six, 
                            X_train_seven),axis=0)
```

```python
y_train_final=np.concatenate((y_train_one,
                            y_train_two,
                            y_train_three,
                            y_train_four,
                            y_train_six,
                            y_train_seven),axis=0)
```

```python
cccv_final.fit(X_train_final, y_train_final)
```

```python
X_valid_final=np.concatenate((X_valid_one,
X_valid_two,
X_valid_three,
X_valid_four,
X_valid_six,
X_valid_seven),axis=0)
```

```python
y_valid_final = np.concatenate((y_valid_one, 
                                 y_valid_two, 
                                 y_valid_three,
                                 y_valid_four, 
                                 y_valid_six, 
                                 y_valid_seven),axis=0)
```

```python
cccv_final_pred =cccv_final.predict(X_valid_final)
```

```python
cccv_final_acc = accuracy_score(y_valid_final, cccv_final_pred)
```

```python
print(cccv_final_acc)
```

```python
from collections import Counter
counter = Counter(cccv_final_pred)
for r,m in counter.items():
    percent = m / len(cccv_final_pred) * 100
    print('Class=%d, n=%d (%.4f%%)' % (r, m, percent))
```
