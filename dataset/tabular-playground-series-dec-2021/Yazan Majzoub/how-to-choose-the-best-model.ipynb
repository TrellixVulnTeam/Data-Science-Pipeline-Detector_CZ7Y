## 1. import libraries and read data

```python
import numpy as np 
import pandas as pd 
import time
from catboost import CatBoostClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
import collections
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score, accuracy_score, classification_report
from collections import Counter
from sklearn.model_selection import KFold, StratifiedKFold
pd.set_option('display.max_columns', None)
#pd.set_option('display.max_rows', None)
train=pd.read_csv('../input/tabular-playground-series-dec-2021/train.csv')
test=pd.read_csv('../input/tabular-playground-series-dec-2021/test.csv')
```

<a id="2."></a>
## 2. drop target and id column


```python
y=train.pop('Cover_Type')
id_=train.pop('Id')
id_test=test.pop('Id')
```

<a id="3."></a>
## 3. Model setting



```python

classifiers = {
    'XGBClassifier':XGBClassifier(),
    "KNearest": KNeighborsClassifier(),
    "Support Vector Classifier": SVC(),
    "DecisionTreeClassifier": DecisionTreeClassifier(),
    #'CatBoostClassifier':CatBoostClassifier() #the best

    

}
```

<a id="4."></a>
## 4. train_test_split


```python
# 5000 a sample (you can change)
from sklearn.model_selection import train_test_split

# This is explicitly used for undersampling.
X_train, X_test, y_train, y_test = train_test_split(train[:5000], y[:5000], test_size=0.2, random_state=42)
```

```python
X_train = X_train.values
X_test = X_test.values
y_train = y_train.values
y_test = y_test.values
```

<a id="5."></a>
## 5. training



```python
from sklearn.model_selection import cross_val_score


for key, classifier in classifiers.items():
    classifier.fit(X_train, y_train)
    training_score = cross_val_score(classifier, X_train, y_train, cv=5)
```

```python

XGBClassifier=XGBClassifier()
a_score = cross_val_score(XGBClassifier, X_train, y_train, cv=5)

KNeighborsClassifier=KNeighborsClassifier()
b_score = cross_val_score(KNeighborsClassifier, X_train, y_train, cv=5)

SVC=SVC()
c_score = cross_val_score(SVC, X_train, y_train, cv=5)

DecisionTreeClassifier=DecisionTreeClassifier()
d_score = cross_val_score(DecisionTreeClassifier, X_train, y_train, cv=5)

```

<a id="6."></a>
## 6. View results



```python
print('XGBClassifier Cross Validation Score', round(a_score.mean() * 100, 2).astype(str) + '%')

print('KNeighborsClassifier Cross Validation Score', round(b_score.mean() * 100, 2).astype(str) + '%')

print('Support Vector Classifier Cross Validation Score', round(c_score.mean() * 100, 2).astype(str) + '%')

print('DecisionTreeClassifier Cross Validation Score', round(d_score.mean() * 100, 2).astype(str) + '%')
```
