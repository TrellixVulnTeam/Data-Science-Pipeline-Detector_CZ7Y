```python
df1=pd.read_csv("../input/publicsunmissionsdecembertps/sub_13folds_num_9.csv")
df2=pd.read_csv("../input/publicsunmissionsdecembertps/submission (12).csv")
df3=pd.read_csv("../input/publicsunmissionsdecembertps/submission (13).csv")
df4=pd.read_csv("../input/publicsunmissionsdecembertps/submission (14).csv")
df5=pd.read_csv("../input/publicsunmissionsdecembertps/submission_final.csv")
df6=pd.read_csv("../input/publicsunmissionsdecembertps/submission_without4.csv")

sub=pd.read_csv("../input/tabular-playground-series-dec-2021/sample_submission.csv")
```

```python
df1.head()
```

```python
from scipy import stats
```

```python
predictions = [df1, df2, df3, df4, df5, df6]
results = pd.DataFrame()
for i, ds in enumerate(predictions):
    results[f'p{i+1}'] = ds['Cover_Type']

print(results.shape)
results.head()
```

```python
sub["Cover_Type"]=stats.mode(np.array(results), axis=1)[0]
```
