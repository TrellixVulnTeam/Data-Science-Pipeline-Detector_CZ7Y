# Import Libraries

```python
import pandas as pd
```

# Load Dataset

```python
# Load dataset
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
        
filename_training          = '/kaggle/input/tabular-playground-series-dec-2021/train.csv'
df           = pd.read_csv(filename_training)
print(df.shape)
df.head()
```

# When you are using AutoViz, you can ask AutoViz to randomly sample a huge data set like this for visualization which makes it fast and easy to use. Notice that I have chosen to sample only about 10% of this data and still get the same insights.

```python
from autoviz.AutoViz_Class import AutoViz_Class

AV = AutoViz_Class()
dftc = AV.AutoViz(
    filename='', 
    sep='' , 
    depVar='Cover_Type', 
    dfte=df, 
    header=0, 
    verbose=1, 
    lowess=False, 
    chart_format='png', 
    max_rows_analyzed=500000, 
    max_cols_analyzed=100
)
```

## Here are the following insights I derived from looking at the charts:
1. Elevation alone can differentiate between the first 4 covertypes and the last one. This might turn out to be the most important variable in this whole dataset!
2. Hillshade_Noon and Hillshade_9am can distinguish between the first 3 covertypes for certain instances and the last 3 covertypes for other instances. It could be very useful variables.
3. Vertical Distance to Hydrology looks like it can do a nice job of separating covertypes as well provided it was visualized as a radial or some other axis.
