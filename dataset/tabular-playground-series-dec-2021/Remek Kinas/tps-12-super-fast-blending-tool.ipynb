```python
import pandas as pd
import numpy as np
from numba import jit

import warnings
warnings.filterwarnings("ignore")
```

```python
sub0 = '../input/tps12-pseudolabels/subs/subs/0.95696.csv'
sub1 = '../input/tps12-pseudolabels/subs/subs/0.95690.csv'
sub2 = '../input/tps12-pseudolabels/subs/subs/0.95687.csv'
sub3 = '../input/tps12-pseudolabels/subs/subs/0.95686.csv'
sub4 = '../input/tps12-pseudolabels/subs/subs/0.95683.csv'
sub6 = '../input/tps12-pseudolabels/subs/subs/0.95673.csv'



result = pd.concat([pd.read_csv(sub0, index_col = "Id"),
                    pd.read_csv(sub1, index_col = "Id"),
                    pd.read_csv(sub2, index_col = "Id"),
                    pd.read_csv(sub3, index_col = "Id"),
                    pd.read_csv(sub4, index_col = "Id"),], axis=1)
```

```python
%%time

@jit(nopython=True)
def np_blending(row):
    return np.argmax(np.bincount(row))

result['Result'] = result.apply(lambda row: np_blending(row.values), axis=1)
```

```python
sub = pd.DataFrame(result['Result']).rename(columns={'Result': 'Cover_Type'})
sub.to_csv("tps12-fast-blend.csv")
sub.tail(3)
```
