```python
import pandas as pd

import warnings
warnings.filterwarnings("ignore")
```

```python
# Take some submissions files (different one). I took 6 but could be more less ... just like in ML ... experiment ...

sub1 = '../input/tps12-pseudolabels/source_submissions/01.csv'
sub2 = '../input/tps12-pseudolabels/source_submissions/02.csv'
sub3 = '../input/tps12-pseudolabels/source_submissions/03.csv'
sub4 = '../input/tps12-pseudolabels/source_submissions/04.csv'
sub5 = '../input/tps12-pseudolabels/source_submissions/04.csv'
sub6 = '../input/tps12-pseudolabels/source_submissions/06.csv'

result = pd.concat([pd.read_csv(sub1, index_col = "Id"),
                    pd.read_csv(sub2, index_col = "Id"),
                    pd.read_csv(sub3, index_col = "Id"),
                    pd.read_csv(sub4, index_col = "Id"),
                    pd.read_csv(sub5, index_col = "Id"),
                    pd.read_csv(sub6, index_col = "Id")], axis=1)

test = pd.read_csv('../input/tabular-playground-series-dec-2021/test.csv', index_col = "Id")
```

```python
def pseudo(row):
    values = row.values
    if (values == values[0]).all():
        return [True, values[0]]
    else:
        return [False, None]

result[['Pseudo', 'Result']] = result.apply(lambda row: pd.Series(pseudo(row)), axis=1)
```

```python
result[result.Pseudo == True]
```

```python
# I take only this rows that every submissions tell the same prediction
pseudo_indexes = result[result.Pseudo].index

pseudo = test[test.index.isin(pseudo_indexes)]

pseudo_classes = result[result.index.isin(pseudo_indexes)]['Result'].astype(int)

pseudo['Cover_Type'] = pseudo_classes
pseudo.to_csv("tps12-pseudolabels.csv")

pseudo.tail(5)
```
