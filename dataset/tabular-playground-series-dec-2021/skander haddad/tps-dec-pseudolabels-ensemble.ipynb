```python
def pseudo(row):
    values = row.values
    if (values == values[0]).all():
        return [True, values[0]]
    else:
        return [False, None]

result[['Pseudo', 'Result']] = result.apply(lambda row: pd.Series(pseudo(row)), axis=1)
```

```python
result[result.Pseudo == True]
```

```python
# I take only this rows that every submissions tell the same prediction
pseudo_indexes = result[result.Pseudo].index

pseudo = test[test.index.isin(pseudo_indexes)]

pseudo_classes = result[result.index.isin(pseudo_indexes)]['Result'].astype(int)

pseudo['Cover_Type'] = pseudo_classes
pseudo.to_csv("tps-dec-pseudolabels.csv")

pseudo.tail(5)
```
