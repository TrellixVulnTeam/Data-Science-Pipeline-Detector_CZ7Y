```python
import numpy as np
import pandas as pd
from supervised.automl import AutoML # MLJAR AutoML
```

```python
train = pd.read_csv("../input/tabular-playground-series-dec-2021/train.csv")
test = pd.read_csv("../input/tabular-playground-series-dec-2021/test.csv")
```

```python
x_cols = train.columns[1:-1].tolist()
y_col = train.columns[-1]
```

```python
automl = AutoML(
    mode="Compete", 
    total_time_limit=10000,
    eval_metric='accuracy',
    algorithms=['Xgboost', 'LightGBM'])
automl.fit(train[x_cols], train[y_col])
```

```python
preds = automl.predict_all(test)
```

```python
preds.head()
```

```python
sub = pd.DataFrame({"Id": test.Id, "Cover_Type": preds['label']})
sub.to_csv("./1_sub.csv", index=False)
```
