```python
p1 = pd.read_csv("../input/tps-12-nn-tpu-pseudolabeling-0-95690/tps12-pseudeo-submission.csv")
p2 = pd.read_csv("../input/tps-12-nn-tpu-pseudolabeling-0-95690/tps12-pseudeo-submission.csv")
p3 = pd.read_csv("../input/tpsdec21-12-eliminate-cover-type-4/submission_without4.csv")
p4 = pd.read_csv("../input/tps202112-reasonable-xgboost-model/submission.csv")
p5 = pd.read_csv("../input/tps202112-reasonable-xgboost-model/submission.csv")
p6 = pd.read_csv("../input/tps202112-reasonable-xgboost-model/submission.csv")
p7 = pd.read_csv("../input/tps-12-super-fast-blending-tool/tps12-fast-blend.csv")
p8 = pd.read_csv("../input/tpsdec21-12-eliminate-cover-type-4/submission_without4.csv")
p9 = pd.read_csv("../input/tpsdec21-12-eliminate-cover-type-4/submission_without4.csv")
p10 = pd.read_csv("../input/tpsdec21-12-eliminate-cover-type-4/submission_without4.csv")

submission = pd.read_csv('../input/tabular-playground-series-dec-2021/sample_submission.csv')

predictions = [p1, p2, p3, p4, p5, p6, p7, p8, p9, p10]
```

```python
results = pd.DataFrame()
for i, ds in enumerate(predictions):
    results[f'p{i+1}'] = ds['Cover_Type']

print(results.shape)
results.head()
```

```python
%%time
results["ensemble"] = stats.mode(np.array(results), axis=1)[0]
results.head()
```

```python
def nunique(a, axis):
    return (np.diff(np.sort(a,axis=axis),axis=axis)!=0).sum(axis=axis)+1
```

```python
results["dif"] = nunique(results.iloc[:,:len(predictions)].values,1) - 1
results.head()
```

```python
results.dif.value_counts()
```

```python
submission['Cover_Type'] = results["ensemble"]
submission.to_csv("submission.csv", index=False)
submission.head()
```

```python
plt.figure(figsize=(10,5))
ax = sns.countplot(x=submission.Cover_Type)
plt.title("Predictions")
plt.xlabel("Cover Type")
ax.bar_label(ax.containers[0])
plt.show()
```
