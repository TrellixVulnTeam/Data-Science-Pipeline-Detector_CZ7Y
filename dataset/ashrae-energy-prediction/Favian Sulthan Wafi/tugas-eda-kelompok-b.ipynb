```python
import pandas as pd # untuk mengimport data
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

**Import Data**

```python
# Data Gedung "gd"
gd = pd.read_csv('../input/ashrae-energy-prediction/building_metadata.csv')
gd.info() #melihat informasi data
```

```python
# Data Energy Gedung variabel train "en_train"
en_train = pd.read_csv('../input/ashrae-energy-prediction/train.csv')
en_train.info()
```

```python
# Data Cuaca variabel train "cu_train"
cu_train = pd.read_csv('../input/ashrae-energy-prediction/weather_train.csv')
cu_train.info()
```

```python
# Data Energy Gedung variabel test "en_test"
en_test = pd.read_csv('../input/ashrae-energy-prediction/test.csv')
en_test.info()
```

```python
# Data Cuaca variabel test "cu_test"
cu_test = pd.read_csv('../input/ashrae-energy-prediction/weather_test.csv')
cu_test.info()
```

Akan digabung data dari variabel train (data gedung, energy gedung, dan data cuaca gedung) dengan menyocokkan building_id pada data gedung dan energy gedung, dan site_id serta timestamp pada data gedung dan cuaca. Lakukan hal yang sama pada variabel test

```python
# Tabel Train (Gabungan Dataset Gedung, Energy Gedung, dan Data Cuaca Gedung pada variabel train)
train = gd.merge(en_train, on ='building_id', how = 'left').merge(cu_train, on = ('site_id', 'timestamp'), how = 'left')
train.head()
```

```python
# Tabel Test (Gabungan Dataset Gedung, Energy Gedung, dan Data Cuaca Gedung pada varaiebl test)
test = gd.merge(en_test, on ='building_id', how = 'left').merge(cu_test, on =('site_id','timestamp'), how ='left')
test.head()
```
