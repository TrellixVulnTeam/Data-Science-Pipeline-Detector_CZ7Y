# Reading the data
## Reading energy comsuption data

```python
df_train_energy = pd.read_csv('/kaggle/input/ashrae-energy-prediction/train.csv', parse_dates=['timestamp'])
df_test_energy = pd.read_csv('/kaggle/input/ashrae-energy-prediction/test.csv', parse_dates=['timestamp'])
```

```python
print("Training data's shape " + str(df_train_energy.shape))
print("Testing data's shape " + str(df_test_energy.shape))
```

The test set is much larger than the training set.

```python
df_train_energy.head(10)
```

```python
df_train_energy.info()
```

```python
df_train_energy.describe()
```

There are 1449 buildings and 4 types of meters: 0: electricity, 1: chilledwater, 2: steam, 3: hotwater

### Disctinct values

```python
df_train_energy.nunique()
```

```python
df_train_energy['meter'] = df_train_energy['meter'].astype('category')
df_test_energy['meter'] = df_train_energy['meter'].astype('category')
```

```python
print("Min date train energy " + str(df_train_energy['timestamp'].min()))
print("Max date train energy " + str(df_train_energy['timestamp'].max()))
print("Min date test energy " + str(df_test_energy['timestamp'].min()))
print("Max date test energy " + str(df_test_energy['timestamp'].max()))
```

As you will see below, the **train** and **test** sests of the **energy usage's** data seem to have the **exact same dates** as the **weather's data**.

### Missing values

```python
df_train_energy.isna().sum()
```

There are **no missing values**, however, not every building has all meter types.
