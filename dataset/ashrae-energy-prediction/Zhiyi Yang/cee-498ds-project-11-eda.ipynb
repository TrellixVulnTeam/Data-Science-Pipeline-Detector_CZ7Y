# 3. Analysis of the data

```python
#the number of buidling in this project
len(building_meta_df.building_id.unique())
```

```python
#primary_use of the buidling in this project
len(building_meta_df.primary_use.unique())
```

```python
plt.figure()
sns.barplot(building_meta_df.primary_use.value_counts().index,building_meta_df.primary_use.value_counts())
plt.xticks(rotation = 90)
plt.ylabel('Quanity of Buildings',fontsize=16)
plt.xlabel('primary_use',fontsize=16)
```

```python
plt.figure()
sns.barplot(building_meta_df.site_id.value_counts().index,building_meta_df.site_id.value_counts())
plt.ylabel('Quanity of Buildings',fontsize=16)
plt.xlabel('Site_id',fontsize=16)
```

```python
plt.figure(figsize=(20,3))
sns.barplot(building_meta_df.year_built.value_counts().index,building_meta_df.year_built.value_counts())
plt.ylabel('Quanity of Buildings',fontsize=16)
plt.xlabel('year_built',fontsize=16)
plt.xticks(rotation = 90)
```

```python
fig,ax=plt.subplots(figsize=(15, 10))
sns.distplot(building_meta_df['square_feet'], color='orange', ax=ax)
plt.xlabel('square_feet',fontsize=16)
plt.show()
```
