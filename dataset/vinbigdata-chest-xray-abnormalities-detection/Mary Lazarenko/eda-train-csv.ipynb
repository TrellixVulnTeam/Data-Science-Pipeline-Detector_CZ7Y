```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import missingno as msno
```

```python
train = pd.read_csv('../input/vinbigdata-chest-xray-abnormalities-detection/train.csv')
train.head()
```

```python
train.rad_id = train.rad_id.apply(lambda x: int(x.replace('R', '')))
train.info()
```

```python
print('\n------------------------------------ UNIQUE VALUES ------------------------------------------\n\n\n')
print("Unique values for class id:",sorted(pd.unique(train.class_id.values)))
print('\n----------------------------------------------------------------------\n')
print("Unique values for class name:",sorted(pd.unique(train.class_name.values)))
print('\n----------------------------------------------------------------------\n')
print("Unique values for rad id:",sorted(pd.unique(train.rad_id.values)))
print('\n\n\n-------------------------------- X AND Y COORDINATES ---------------------------------------\n\n\n')
print('Mean value of minimum X coordinate of the objects bounding box is',np.mean(train.x_min), ', mean value of maximum X is',np.mean(train.x_max))
print('Mean value of minimum Y coordinate of the objects bounding box is',np.mean(train.y_min), ', mean value of maximum Y is',np.mean(train.y_max))
print('\n----------------------------------------------------------------------\n')
print('Values of minimum X coordinates lie between', min(train.x_min.dropna()), 'and', max(train.x_min.dropna()))
print('Values of maximum X coordinates lie between', min(train.x_max.dropna()), 'and', max(train.x_max.dropna()))
print('\n----------------------------------------------------------------------\n')
print('Values of minimum Y coordinates lie between', min(train.y_min.dropna()), 'and', max(train.y_min.dropna()))
print('Values of maximum Y coordinates lie between', min(train.y_max.dropna()), 'and', max(train.y_max.dropna()))
print('\n\n\n-------------------------------- MOST FREQUENT VALUES --------------------------------------\n\n\n')
print('The most frequent value in rad id is:', train.rad_id.value_counts().idxmax())
print('The most frequent value in class name is:', train.class_name.value_counts().idxmax())
print('The most frequent value in class id is:', train.class_id.value_counts().idxmax(),'\n')
```

```python
print('\n----------------- X min ----------------------\n')
print(train.x_min.describe())
print('\n----------------- X max ----------------------\n')
print(train.x_max.describe())
print('\n----------------- Y min ----------------------\n')
print(train.y_min.describe())
print('\n----------------- Y max ----------------------\n')
print(train.y_max.describe())
```

```python
fig, axes = plt.subplots(figsize = (25,7), nrows=1, ncols=3)
ax0, ax1, ax2 = axes.flatten()

ax0.hist(train.class_id, bins=35, color = "orchid")
ax0.set_xticks(pd.unique(train.class_id.values))
ax0.set_title('class id', fontsize=22)

ax1.hist(train.rad_id, bins=35, color = "mediumorchid")
ax1.set_xticks(pd.unique(train.rad_id.values))
ax1.set_title('rad id', fontsize=22)

ax2.hist(train.class_name, bins=35, color = "darkorchid")
ax2.set_xticklabels(pd.unique(train.class_name.values), rotation='vertical', fontsize=14)
ax2.set_title('class name', fontsize=22)
```

```python
fig, axes = plt.subplots(figsize = (20,10), nrows=2, ncols=2)
ax0, ax1, ax2, ax3 = axes.flatten()

ax0.hist(train.x_min, bins=55, color = "skyblue")
ax0.axvline(x=602, color='royalblue', linestyle='dashed', linewidth=2)
ax0.axvline(x=1457, color='royalblue', linestyle='dashed', linewidth=2)
ax0.axvline(x=1014, color='cornflowerblue', linewidth=2)
ax0.set_title('X minimum', fontsize=18)

ax1.hist(train.x_max, bins=55, color = "skyblue")
ax1.axvline(x=1010, color='royalblue', linestyle='dashed', linewidth=2)
ax1.axvline(x=1567, color='cornflowerblue', linewidth=2)
ax1.axvline(x=1947, color='royalblue', linestyle='dashed', linewidth=2)
ax1.set_title('X maximum', fontsize=18)

ax2.hist(train.y_min, bins=55)
ax2.axvline(x=627, color='orchid', linestyle='dashed', linewidth=2)
ax2.axvline(x=935, color='cornflowerblue', linewidth=2)
ax2.axvline(x=1471, color='orchid', linestyle='dashed', linewidth=2)
ax2.set_title('Y minimum', fontsize=18)

ax3.hist(train.y_max, bins=55)
ax3.axvline(x=1009, color='orchid', linestyle='dashed', linewidth=2)
ax3.axvline(x=1411, color='cornflowerblue', linewidth=2)
ax3.axvline(x=1911, color='orchid', linestyle='dashed', linewidth=2)
ax3.set_title('Y maximum', fontsize=18)
```
