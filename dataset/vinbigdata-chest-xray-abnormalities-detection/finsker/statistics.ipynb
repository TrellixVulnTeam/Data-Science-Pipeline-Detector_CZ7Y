```python
class index:
    def __init__(self, xmax, ymax):
        self.xmax = xmax
        self.ymax = ymax
        self.i = 0
        self.x = 0
        self.y = 0
        
    def inc(self):
        self.i += 1
        if self.x != self.xmax - 1:
            self.x+=1
        else:
            self.x=0
            self.y+=1
```

```python
xmax, ymax = (6,3) 
ind = index(xmax, ymax)
fig, axes = plt.subplots(xmax, ymax, figsize=(24,24), constrained_layout=True)
plt.xticks(fontsize = 14, rotation = 90)
plt.yticks(fontsize = 14)
# fig.tight_layout()

for rad_id in sorted(train_csv['rad_id'].unique()):
    filtered_ids = train_csv.query(f'rad_id == "{rad_id}"')
    sns.countplot(filtered_ids.class_name, ax = axes[ind.x, ind.y])
    axes[ind.x, ind.y].set_title(label=f'{rad_id}')
    axes[ind.x, ind.y].tick_params(labelrotation=90)
    ind.inc()
```
