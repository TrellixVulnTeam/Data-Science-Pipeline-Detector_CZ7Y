```python
import pandas as pd
from sklearn import model_selection
```

```python
df = pd.read_csv("../input/vinbigdata-chest-xray-abnormalities-detection/train.csv")
```

```python
df.head()
```

```python
df["kfold"] = -1
df = df.sample(frac=1).reset_index(drop=True)
y = df.class_id.values
```

```python
kf = model_selection.GroupKFold(n_splits=5)

# I group by image_id. Should we group by rad_id? ðŸ¤”
for f, (t_, v_) in enumerate(kf.split(X=df, y=y, groups=df.image_id.values)):
    df.loc[v_, "kfold"] = f
```

```python
df.to_csv("train_folds.csv", index=False)
print('hopefully my train_fold will help you to build good model.')
```
