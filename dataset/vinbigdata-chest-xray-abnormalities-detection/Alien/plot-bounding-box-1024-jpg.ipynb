```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import cv2
import math
```

```python
df = pd.read_csv('../input/vinbigdata-1024-jpg-dataset/train.csv')
df_class_id =  pd.unique(df['image_id'])
print(df_class_id.shape)
```

```python
save_dir_train_diseases = '/kaggle/tmp/diseases/'
save_dir_train_normal = '/kaggle/tmp/normal/'
save_dir_test = '/kaggle/tmp/test/'

os.makedirs(save_dir_train_diseases, exist_ok=True)
os.makedirs(save_dir_train_normal, exist_ok=True)
os.makedirs(save_dir_test, exist_ok=True)

```

```python
border_color_list = [(255,0,0), (0,255,0), (0,0,255), (255,255,0), (0,255,255), (255,0,255), 
                     (125,0,0),(0,125,0), (0,0,125), (125,125,0), (0,125,125), (189,252,201)
                    , (180,77,250), (221,160,221)]
for i in range(15000):
    image_id = df_class_id[i]
    img = cv2.imread(f'../input/vinbigdata-1024-jpg-dataset/train/{image_id}.jpg')
    df_img = df[(df.loc[:,'image_id'] == image_id)].reset_index(drop=True)

    for j in range(df_img.shape[0]):

        if df_img .loc[j,'class_id'] == 14:
            continue

        for k in range(14):        
            if df_img .loc[j,'class_id'] == k:
                border_color = border_color_list[k]

        a = df_img.loc[j,'x_min']
        b = df_img.loc[j,'y_min']
        c = df_img.loc[j,'x_max']
        d = df_img.loc[j,'y_max']
        cv2.rectangle(img, (int(a), int(b)), (int(c), int(d)), border_color, 2)
    for j in range(df_img.shape[0]):
        if df_img .loc[j,'class_id'] != 14:
            cv2.imwrite(save_dir_train_diseases +image_id+'.jpg',img)
            break
        if j == df_img.shape[0] - 1:
            if df_img .loc[df_img.shape[0]-1,'class_id'] == 14:
                cv2.imwrite(save_dir_train_normal +image_id+'.jpg',img)

!tar -zcf diseases.tar.gz -C "/kaggle/tmp/diseases/" .
!tar -zcf normal.tar.gz -C "/kaggle/tmp/normal/" .
```
