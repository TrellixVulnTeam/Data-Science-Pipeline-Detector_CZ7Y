```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pydicom

import warnings
warnings.filterwarnings('ignore')
```

> ## 1. Dicom to Numpy Array:

```python
def dcm_to_arr(root_dir, no_of_imgs=10):
    dcm_files = []
    dcm_to_arr = []
    for dirs, sub_dirs, files in os.walk(root_dir):
        for file in files:
            if '.dicom' in file.lower():
                dcm_files.append(os.path.join(dirs,file))

    for dcm_file in dcm_files[:no_of_imgs]:
        # Get ref file
        dcm_to_arr.append(pydicom.read_file(dcm_file).pixel_array)
    return dcm_files, dcm_to_arr
```

```python
dcm_files, images = dcm_to_arr(root_dir='/kaggle/input')
```
