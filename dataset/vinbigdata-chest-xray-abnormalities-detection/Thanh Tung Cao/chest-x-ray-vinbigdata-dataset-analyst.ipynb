```python
import os
import numpy as np # linear algebra

import matplotlib.pyplot as plt
import pydicom

import pandas as pd
```

# Utils function

```python
def plot_pixel_array(image, figsize=(10,10)):
    '''
        param: 
            - image: np.array
    '''
    plt.figure(figsize=figsize)
    plt.imshow(image, cmap=plt.cm.bone)
    plt.show()
    
def show_dcm_info(patientImg):
    '''
        param: 
            - patientImg: header data of dicom file.
    '''
    print("File Path:", filePath)
    print("Patient's Gender :", patientImg.PatientSex)
    
    if 'PixelData' in patientImg:
        rows = int(patientImg.Rows)
        cols = int(patientImg.Columns)
        print("Image size : {rows:d} x {cols:d}, {size:d} bytes".format(
            rows=rows, cols=cols, size=len(patientImg.PixelData)))
        # Some patient have no pixel spacing value
        if 'PixelSpacing' in patientImg:
            print("Pixel spacing :", patientImg.PixelSpacing)
```

```python
DIR = '/kaggle/input/vinbigdata-chest-xray-abnormalities-detection'
trainPath: list = [(DIR + '/train/' + x) for x in os.listdir(DIR + '/train')]
testPath: list = [(DIR + '/test/' + x) for x in os.listdir(DIR + '/test')]
print(os.listdir(DIR))
```

## Show information of dataset

```python
sampleId: int = np.random.randint(0, len(trainPath))
dicom_info = pydicom.dcmread(trainPath[sampleId])
dicom_array = dicom_info.pixel_array

print('the number of train set: {}'.format(len(trainPath)))
print('the number of test set: {}'.format(len(testPath)))
print('Patient #{}'.format(sampleId))
print('Type of image:', type(dicom_array))
print('Shape of image:', dicom_array.shape)
plt.figure(figsize=(7, 7))
plt.imshow(dicom_array, cmap=plt.cm.bone)
plt.show()
```

## Show some patient information

```python
i = 1
num_to_plot = 2
for filePath in trainPath:
    data_info = pydicom.dcmread(filePath)
    show_dcm_info(data_info)
    plot_pixel_array(data_info.pixel_array, (5,5))
    if i >= num_to_plot:
        break
    i += 1
```
