**Plotting**
>Plot xray with bounding boxes

```python
bbox_classes=[]
image_ids=train_df['image_id'].unique()
columns=['class_name','x_min','y_min','x_max','y_max']
for id_ in tqdm(image_ids):
    bbox_classes.append(train_df[columns][train_df['image_id']==id_].values)
```

```python
def plot_xray(image_id,bbox):
    global class_names
    class_names=list(train_df['class_name'].unique())
    colors = plt.cm.hsv(np.linspace(0, 1, len(class_names))).tolist()
    fig=plt.figure(figsize=(10,10))
    current_axis = plt.gca()
    img=pydicom.read_file(os.path.join(train_path,image_id+'.dicom')).pixel_array
    plt.imshow(img,cmap='gray')
    for box in bbox:
        xmin=box[1]
        ymin=box[2]
        xmax=box[3]
        ymax=box[4]
        label=box[0]
        color = colors[class_names.index(label)]
        current_axis.add_patch(plt.Rectangle((xmin, ymin), xmax-xmin, ymax-ymin, 
                                             color=color, fill=False, linewidth=2))
        
        current_axis.text(xmin, ymin, label, size='x-large', 
                          color='white', bbox={'facecolor':'green', 'alpha':1.0})
```

```python
plot_xray(image_ids[8],bbox_classes[8])
plot_xray(image_ids[6],bbox_classes[6])
```
