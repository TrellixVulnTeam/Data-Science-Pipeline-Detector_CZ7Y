```python
# reference for normalization: https://www.kaggle.com/raddar/convert-dicom-to-np-array-the-correct-way

def load_data(file_path, fix_monochrome = True):
    
    # read the dicom file
    RefDs = pydicom.read_file(file_path)
    
    
    # Load dimensions based on the number of rows, columns,
    x_dim,y_dim = (int(RefDs.Rows), int(RefDs.Columns))
    
    # load pixel value
    xray = RefDs.pixel_array
    
    # flip the pixels upside-down so that image is not inverted
    xray = np.flipud(xray)
    
    if fix_monochrome and RefDs.PhotometricInterpretation == "MONOCHROME1":
        xray = np.amax(xray) - xray
        
    xray = xray - np.min(xray)
    xray = xray / np.max(xray)
    xray = (xray * 255).astype(np.uint8)
        
    return xray, y_dim, x_dim
    
    
```

```python
def show_image(img_data, x_dim, y_dim):
    pyplot.figure(dpi=100)
    pyplot.axes().set_aspect('equal')
    pyplot.set_cmap(pyplot.gray())
    pyplot.pcolormesh(np.arange(x_dim),np.arange(y_dim),img_data)
```

```python
# load and show tha image
show_image(*load_data(path + 'train/0108949daa13dc94634a7d650a05c0bb.dicom'))
```
