```python
import os 
import cv2
import glob
import imageio
import numpy as np
from tqdm import tqdm
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
TOMAKE = 1024 #512, 256
PATH = '../input/vinbigdata-{}-image-dataset/vinbigdata/'.format(TOMAKE)
TRAIN_FILES, TEST_FILES = PATH + 'train', PATH + 'test'


train_files = glob.glob(TRAIN_FILES + '/*.png')
test_files = glob.glob(TEST_FILES + '/*.png')
csv_files = glob.glob(PATH + '/*.csv')
```

```python
def clahe(files):
    CLIP_LIMIT=2.0
    GRID_SIZE = (10, 10)
    clahe = cv2.createCLAHE(clipLimit=CLIP_LIMIT, tileGridSize=GRID_SIZE)
    tosave = [os.path.join("vinbigdata", *file.split('/')[-2:]) for file in files]
    for reads, saves in tqdm(zip(files, tosave)):
        temp = cv2.imread(reads, 0)
        temp2 = clahe.apply(temp)
        temp2 = cv2.cvtColor(temp2, cv2.COLOR_GRAY2BGR)
        imageio.imwrite(saves, temp2)
    

def copy_csv(files):
    from shutil import copyfile
    for src in files:
        copyfile(src, 'vinbigdata/'+ src.split('/')[-1])
```

```python
if not os.path.exists('vinbigdata'):
    os.mkdir('vinbigdata')
    os.mkdir('vinbigdata/train')
    os.mkdir('vinbigdata/test')
```

```python
clahe(train_files)
clahe(test_files)
```

```python
copy_csv(csv_files)
```
