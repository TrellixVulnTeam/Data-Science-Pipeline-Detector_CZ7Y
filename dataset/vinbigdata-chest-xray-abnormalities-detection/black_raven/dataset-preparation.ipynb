```python
%%time
import numpy as np 
import pandas as pd 
import os
import cv2
from utilities_x_ray import read_xray
from tqdm import tqdm
import shutil
import warnings
warnings.filterwarnings("ignore")

print('Preparing numpy files from dicom: ')
path = '../input/vinbigdata-chest-xray-abnormalities-detection/train/'
if os.path.isdir('./images/'):
    shutil.rmtree('./images/')
    
os.mkdir('./images/')
os.mkdir('./images/train/')
os.mkdir('./images/test/')
print('Train set')
for img in tqdm(os.listdir(path)):
    im_name = img.split('.dicom')[0]
    image = read_xray(path+img)
    image = cv2.resize(image,(256,256),cv2.INTER_AREA)
    image = np.expand_dims(image,axis=2)
    np.save(f'./images/train/{im_name}.npy',image)

path = '../input/vinbigdata-chest-xray-abnormalities-detection/test/'
print('Test set')
for img in tqdm(os.listdir(path)):
    im_name = img.split('.dicom')[0]
    image = read_xray(path+img)
    image = cv2.resize(image,(256,256),cv2.INTER_AREA)
    image = np.expand_dims(image,axis=2)
    np.save(f'./images/test/{im_name}.npy',image)
```
