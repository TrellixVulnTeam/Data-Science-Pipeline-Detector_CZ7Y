```python
# import libraries 

import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import figure
import pydicom
from joblib import Parallel, delayed
from tqdm import tqdm_notebook as tqdm
import cv2
from os.path import isfile, join

%matplotlib inline
```

```python
train_data_frame = pd.read_csv('../input/vinbigdata-chest-xray-abnormalities-detection/train.csv')
sample_submission =  pd.read_csv("../input/vinbigdata-chest-xray-abnormalities-detection/sample_submission.csv")

print(train_data_frame["class_name"].unique())
print(len(train_data_frame["class_name"].unique()))
```

```python
sns.countplot(x="class_id", data=train_data_frame)
```

It's all well and good knowing the class IDs but now let's look at the class count under their respective names. 

```python
plt.figure(figsize=(30,8))
sns.countplot(x="class_name", data=train_data_frame)
plt.show()
```

It appears that no finding is the majority class, where pneumothorax has the fewest entries. 

```python
plt.figure(figsize=(10, 10))
sns.pairplot(train_data_frame, hue='class_id')
plt.show()
```

```python
train_images_dir = '../input/vinbigdata-chest-xray-abnormalities-detection/train/'
train_images = [f for f in os.listdir(train_images_dir) if isfile(join(train_images_dir, f))]
plt.style.use('default')
fig=plt.figure(figsize=(20, 10))
columns = 8; rows = 4
for i in range(1, columns*rows +1):
    ds = pydicom.dcmread(train_images_dir + train_images[i])
    fig.add_subplot(rows, columns, i)
    plt.imshow(ds.pixel_array, cmap=plt.cm.bone)
    fig.add_subplot
```
