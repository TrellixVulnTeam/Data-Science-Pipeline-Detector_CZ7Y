```python
#importing import packages
import os
import cv2
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pydicom
from PIL import Image
```

```python
train_df=pd.read_csv('/kaggle/input/vinbigdata-chest-xray-abnormalities-detection/train.csv')##Reading the Training Files
train_df.sample(10)
```

```python
print("Total_Unique Class in Train::",train_df.class_name.nunique())
```

```python
## Countinng the class training sample ###
print("Total Class is",len(train_df['class_name'].unique()))
for class_name in train_df['class_name'].unique():
    print("Defect type is ",class_name, " and number of training set in train is",len(train_df[train_df['class_name']==class_name]))
```

```python
#####ploting the bar chart for class ######
class_name=list()
total_count=list()
for name in train_df['class_name'].unique():
    class_name.append(name)
    total_count.append(len(train_df[train_df['class_name']==name]))
plt.figure(figsize=(15,5))
sns.set(font_scale = 1.5)
ax = sns.barplot(y=class_name, x=total_count)
```
