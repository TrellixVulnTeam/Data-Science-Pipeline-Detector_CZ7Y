```python
num_records = 40428967
import random
import numpy as np
sample_size = 2000000
skip_values = sorted(random.sample(range(1,num_records), num_records - sample_size))
```

```python
import numpy as np
types_train = {
    'id': np.dtype(int),
    'click': np.dtype(int),
     'hour': np.str,
    'C1': np.dtype(str),
    'banner_pos': np.dtype(str),
    'site_id': np.dtype(str),
    'site_domain': np.dtype(str), 
    'site_category': np.dtype(str),
    'app_id': np.dtype(str),
    'app_domain': np.dtype(str),
    'app_category': np.dtype(str),
    'device_id': np.dtype(str),
    'device_ip': np.dtype(str),
    'device_model': np.dtype(str),
    'device_type': np.dtype(str),
    'device_conn_type': np.dtype(str),
    'C14': np.dtype(str),
    'C15': np.dtype(str),
    'C16': np.dtype(str),
    'C17': np.dtype(str),
    'C18': np.dtype(str),
    'C19': np.dtype(str),
    'C20': np.dtype(str),
    'C21':np.dtype(str)
}
```

```python
import pandas as pd
pd.set_option('display.max_columns', None)
parse_date = lambda val : pd.datetime.strptime(val, '%y%m%d%H')
train = pd.read_csv("../input/avazu-ctr-prediction/train.gz", parse_dates=['hour'], date_parser=parse_date, dtype=types_train, skiprows=skip_values)
train.to_csv("./2Mrecords.csv",index=False)
train.head()
```

```python
train['click'].value_counts()
```

```python
click_0=train[train['click']==0]
click_0.head()
```

```python
click_0=click_0.sample(n=50000)
click_0.reset_index(inplace=True,drop=True) 
click_0.head()
```

```python
click_1=train[train['click']==1]
click_1.head()
```

```python
click_1=click_1.sample(n=50000)
click_1.reset_index(inplace=True,drop=True) 
click_1.head()
```

```python
train=pd.concat([click_1,click_0])
train=train.sample(frac =1)
train.reset_index(inplace=True,drop=True) 
train.head()
```
