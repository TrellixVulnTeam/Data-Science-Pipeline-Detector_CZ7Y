```python
from pycaret.classification import *
```

```python
final_model = load_model('../input/ctr-prediction-model/CTRPredictionmodel')
```

```python
import numpy as np
types_train = {
    'id': np.dtype(str),
    'click': np.dtype(int),
    'hour': np.dtype(int),
    'C1': np.dtype(str),
    'banner_pos': np.dtype(str),
    'site_id': np.dtype(str),
    'site_domain': np.dtype(str), 
    'site_category': np.dtype(str),
    'app_id': np.dtype(str),
    'app_domain': np.dtype(str),
    'app_category': np.dtype(str),
    'device_id': np.dtype(str),
    'device_ip': np.dtype(str),
    'device_model': np.dtype(str),
    'device_type': np.dtype(str),
    'device_conn_type': np.dtype(str),
    'C14': np.dtype(str),
    'C15': np.dtype(str),
    'C16': np.dtype(str),
    'C17': np.dtype(str),
    'C18': np.dtype(str),
    'C19': np.dtype(str),
    'C20': np.dtype(str),
    'C21':np.dtype(str)
}
```

```python
parse_date = lambda val : pd.datetime.strptime(val, '%y%m%d%H')
unseen_data = pd.read_csv('../input/test-data-for-ctr-prediction/ctr_testdata15.csv', parse_dates = ['hour'],
                          date_parser = parse_date, dtype=types_train)
unseen_data.head()
```

```python
unseen_data.hour.describe()
```

```python
predictions = predict_model(final_model, data = unseen_data)
predictions.head()
```

```python
predictions=predictions[['id','Label']]
predictions.head()
```

```python
predictions.rename({'Label': 'click'}, axis=1, inplace=True)
predictions.head()
```

```python
predictions.click.value_counts()
```
