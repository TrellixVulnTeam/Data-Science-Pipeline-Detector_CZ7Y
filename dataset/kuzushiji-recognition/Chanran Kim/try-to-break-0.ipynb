```python
for i, row in df_submission.iterrows():
    samples = df_train.iloc[i%df_train.shape[0]]['labels']
    result = ''
    tmp = ''
    j=0
    for sample in samples.split(' '):
        if j == 0:
            tmp += sample + ' '
            j += 1
        elif j == 1:
            tmp += sample + ' '
            j += 1
        elif j == 2:
            tmp += sample + ' '
            j += 1
        elif j == 3:
            j += 1
        else:
            result += tmp
            tmp = ''
            j = 0

    if len(result) == 0:
        df_submission.set_value(i, 'labels', '')
    else:
        df_submission.set_value(i, 'labels', result[:-1])
```
