### Parse and visualize image data

```python
import re
```

```python
sample = data.iloc[500]

labels = sample.labels
labels = re.findall(r"U\+.{4} \d+ \d+ \d+ \d+",labels)
```

```python
img = cv2.imread(TRAIN_IMAGES + f'{sample.image_id}.jpg')
```

```python
symbol_idx = 0 
x_idx = 1
y_idx = 2
w_idx = 3
h_idx = 4
for label in labels:
    values = label.split()
    x = int(values[x_idx])
    y = int(values[y_idx])
    w = int(values[w_idx])
    h = int(values[h_idx])
    
    cv2.rectangle(img,(x,y), (x+w,y+h),(0,0,255), 10)
    
```
