# Importing data and libraries all I need

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

# Read data

```python
train = pd.read_csv("/kaggle/input/covid19-global-forecasting-week-4/train.csv") 
test = pd.read_csv("/kaggle/input/covid19-global-forecasting-week-4/test.csv") 
```

# confirmed_value vs dead_Value
# with scikit learn Linear Regression

```python
confirmed_value = train.ConfirmedCases.values.reshape(-1,1)
dead_value = train.Fatalities.values.reshape(-1,1)

regression = LinearRegression()
regression.fit(confirmed_value,dead_value)

print(regression.predict([[4000000]]).reshape(-1,1))

print(regression.predict([[3511229]]).reshape(-1,1))
plt.scatter(train.ConfirmedCases,train.Fatalities)

x = np.arange(min(train.ConfirmedCases),max(train.ConfirmedCases)).reshape(-1,1)

plt.plot(x,regression.predict(x),color="r")

plt.xlabel('Confirmed')
plt.ylabel('Deaths')
plt.show()
```

# scikit learn Linear Regression ACCURACY
# %80 accuracy

```python
print(r2_score(dead_value,regression.predict(confirmed_value)))
# 0.80 very good accuracy i think
```

# Polynomial Regression
# Confirmed vs Deaths

```python
Deaths_data = train.iloc[:,5].values.reshape(-1,1)

Confirmed_data = train.iloc[:,4].values.reshape(-1,1)
```

```python
Poly_regression = PolynomialFeatures(degree = 4)

Deaths_data_poly = Poly_regression.fit_transform(Deaths_data)
```

```python
regression2 = LinearRegression() 
regression2.fit(Deaths_data_poly,Confirmed_data)
```

```python
plt.scatter(Deaths_data,Confirmed_data,color='red')
plt.plot(Deaths_data,regression2.predict(Deaths_data_poly))

plt.show()
```

```python
plt.plot(Deaths_data,regression2.predict(Deaths_data_poly))

plt.show()
```

# Polynomial Regression Model ACCURACY
# %88 accuracy

```python
print(r2_score(Confirmed_data,regression2.predict(Deaths_data_poly)))
```

# Desicion Tree model

```python
Deaths_data_value = train.iloc[:,5].values.reshape(-1,1)

Confirmed_data_value = train.iloc[:,4].values.reshape(-1,1)
```

```python
Desicion_regression = DecisionTreeRegressor()
Desicion_regression.fit(Deaths_data_value,Confirmed_data_value)
```

```python
# this means
# 1000 dead in 24551 confirmed data 
print(Desicion_regression.predict(np.array([[1000]])))
```

# Normal plot

```python
plt.scatter(Deaths_data_value,Confirmed_data_value,color='red')
```

# Normal with Desicion Tree model

```python
x = np.arange(min(Deaths_data_value),max(Deaths_data_value),0.01).reshape(-1,1)

plt.scatter(Deaths_data_value,Confirmed_data_value,color='red')
plt.plot(x,Desicion_regression.predict(x),color='blue')
plt.xlabel('Deaths_data_value')
plt.ylabel('Confirmed_data_value')
plt.title('Decision tree model')
plt.show()
```

# only Desicion Tree model

```python
plt.plot(x,Desicion_regression.predict(x),color='blue')
plt.xlabel('Deaths_data_value')
plt.ylabel('Confirmed_data_value')
plt.title('Decision tree model')
plt.show()
```

# Desicion Tree model ACCURACYÂ¶
# %99 accuracy

```python
print(r2_score(Confirmed_data_value,Desicion_regression.predict(Deaths_data_value)))
```

# RandomForestRegressor model

```python
Deaths_data_value = train.iloc[:,5].values.reshape(-1,1)

Confirmed_data_value = train.iloc[:,4].values.reshape(-1,1)

R_regression = RandomForestRegressor(n_estimators=10,random_state=0)


R_regression.fit(Deaths_data_value,Confirmed_data_value)

print(R_regression.predict(np.array([[1000]])))
```

# CHECK THE ACCURACY
# %96 accuracy 

```python
print(r2_score(Confirmed_data_value,R_regression.predict(Deaths_data_value)))
```
