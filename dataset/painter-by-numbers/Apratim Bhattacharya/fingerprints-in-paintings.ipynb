```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from PIL import Image, ImageFilter
import random
import cv2
import os, glob

#t = pd.read_csv('../input/train_info.csv'); t.head()
#s = pd.read_csv('../input/submission_info.csv'); s.head()
train_files = [f for f in glob.glob("../input/train_2/*")]
i_ = 0
plt.rcParams['figure.figsize'] = (10.0, 10.0)
plt.subplots_adjust(wspace=0, hspace=0)
for l in train_files[:100]:
    im = cv2.imread(l)
    im = cv2.resize(im, (50, 50)) 
    plt.subplot(10, 10, i_+1) #.set_title(l)
    plt.imshow(cv2.cvtColor(im, cv2.COLOR_BGR2RGB)); plt.axis('off')
    i_ += 1
```

## Artwork from Artwork - Is that Possible?

```python
im1 = Image.open('../input/train_2/23504.jpg')
im2 = Image.open('../input/train_2/22873.jpg')
w1, h1 = im1.size; w2, h2 = im2.size
p2 = im2.load() #get pixels
for x in range(0, w1,2):
    if x < w2:
        for y in range(0, h1,2):
            if y < h2:
                 im1.putpixel((x, y), p2[x,y])
plt.imshow(im1); plt.axis('off')
```

```python
iFilters = [ImageFilter.BLUR, ImageFilter.CONTOUR, ImageFilter.DETAIL, 
            ImageFilter.EDGE_ENHANCE, ImageFilter.EDGE_ENHANCE_MORE, ImageFilter.EMBOSS, 
            ImageFilter.FIND_EDGES, ImageFilter.SMOOTH, ImageFilter.SMOOTH_MORE, 
            ImageFilter.SHARPEN]
plt.rcParams['figure.figsize'] = (6.0, 20.0)
plt.subplots_adjust(wspace=0, hspace=0)
for i in range(10):
    im3 = im2
    im3 = im3.filter(iFilters[i])
    #im = cv2.resize(im, (300, 300)) 
    plt.subplot(5, 2, i+1) #.set_title(l)
    plt.imshow(im3); plt.axis('off')
```

## Lets add a third image

```python
plt.rcParams['figure.figsize'] = (10.0, 10.0)
im2 = Image.open(train_files[61])
im2 = im2.resize((w1,h1), Image.ANTIALIAS)
w2, h2 = im2.size
p1 = im1.load(); p2 = im2.load()
for x in range(0, w1,1):
    if x < w2:
        for y in range(0, h1,1):
            if y < h2:
                 im1.putpixel((x, y), (p1[x,y][0], p1[x,y][1], p2[x,y][2]))
plt.imshow(im1); plt.axis('off')
```
