```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
```

# Mining rate

The exact formula to get the % mined is ln(num_ships_in_fleet) / 20

```python
x = []
y = []

for i in range(1, 404):
    x.append(i)
    y.append(np.log(i) / 20)
plt.title('Mining rate')
plt.plot(x, y)
```

```python
plt.title('Slope of Mining rate')
plt.plot(x[1:], [now - prev for prev, now in zip(y, y[1:])])
```

```python
plt.title('Mining rate')
plt.plot(x[:30], y[:30])
```

# The Downside of smaller fleets

Observant readers will notice these are the Fibbonaci numbers, and and this is given by the formula floor(2 * ln(num_ships)) + 1

```python
x = []
y = []

for i in range(1, 404):
    x.append(i)
    y.append(np.floor(np.log(i) *2) + 1)
    
    
plt.title('The Downside of smaller fleets')
plt.plot(x, y)
```

```python
plt.title('The Downside of smaller fleets')
plt.plot(x[:50], y[:50])
```
