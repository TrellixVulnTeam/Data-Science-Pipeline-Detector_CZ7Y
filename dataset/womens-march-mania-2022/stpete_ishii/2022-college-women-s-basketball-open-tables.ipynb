# 2022 College Women's Basketball Open Tables

```python
import numpy as np
import pandas as pd
import os
```

```python
paths=[]
names=[]
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        paths+=[os.path.join(dirname, filename)]
        names+=[filename[0:-4]]
```

```python
for i in range(len(paths)):
    names[i]=pd.read_csv(paths[i],encoding='cp932')
    print(paths[i].split('/')[-1][0:-4])
    display(names[i])
    print()
```
