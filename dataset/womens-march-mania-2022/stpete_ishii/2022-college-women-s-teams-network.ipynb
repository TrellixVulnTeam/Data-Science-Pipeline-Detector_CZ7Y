```python
import numpy as np
import pandas as pd
import os
from tqdm import tqdm
import matplotlib.pyplot as plt
import networkx as nx
```

```python
paths=[]
names=[]
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        paths+=[os.path.join(dirname, filename)]
        names+=[filename[0:-4]]
```

```python
for i in range(len(paths)):
    names[i]=pd.read_csv(paths[i],encoding='cp932')
    print(str(i)+'. '+paths[i].split('/')[-1][0:-4])
    #display(names[i])
    #print()
```

```python
A1=names[3]
display(A1)
```

```python
A2=names[8]
display(A2)
```

```python
AA=pd.concat([A1,A2],axis=0)
AA['text']=AA['WTeamID'].astype(str)+' '+AA['LTeamID'].astype(str)
display(AA)
AA1=AA[AA['Season']==1998]
AA2=AA1[['text']]
```

```python
AA2.to_csv('edgelist.txt',index=False,header=None)
```

```python
G = nx.read_edgelist('./edgelist.txt', nodetype=str)
plt.figure(figsize=(16,12))
pos = nx.spring_layout(G, k=1.0) #k:denseness
nx.draw_networkx_edges(G, pos, edge_color='y')
nx.draw_networkx_nodes(G, pos, node_color='r', alpha=0.3) #alpha:transparency
nx.draw_networkx_labels(G, pos, font_size=10)
plt.axis('off')
plt.show()
```
