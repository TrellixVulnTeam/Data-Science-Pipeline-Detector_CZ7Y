# 2022 College Women's Team Ratio

```python
import numpy as np
import pandas as pd
import os
from tqdm import tqdm
```

```python
paths=[]
names=[]
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        paths+=[os.path.join(dirname, filename)]
        names+=[filename[0:-4]]
```

```python
for i in range(len(paths)):
    names[i]=pd.read_csv(paths[i],encoding='cp932')
    print(str(i)+'. '+paths[i].split('/')[-1][0:-4])
    #display(names[i])
    #print()
```

```python
A=names[3]
display(A)
```

```python
A2=names[8]
display(A2)
```

```python
B=names[2]
display(B)
Year=[]
Team1=[]
Team2=[]
for item in B['ID']:
    item2=item.split('_')
    Year+=[int(item2[0])]
    Team1+=[int(item2[1])]
    Team2+=[int(item2[2])]
B['Year']=Year
B['Team1']=Team1
B['Team2']=Team2
display(B)
```

```python
Teams=list(set(Team1)|set(Team2))
print(len(Teams))
print(Teams[0:3])
print(Teams[-3:])
```

```python
C=pd.DataFrame(columns=['Team','Win','Lose','Ratio'])
C['Team']=Teams
C.loc[:,'Win']=0
C.loc[:,'Lose']=0
display(C)
```

```python
for i in tqdm(range(len(A))):
    for j in range(len(C)):
        if C.iloc[j,0]==A.iloc[i,2]:
            C.iloc[j,1]+=1
        elif C.iloc[j,0]==A.iloc[i,4]:
            C.iloc[j,2]+=1
        else:
            pass
        
for i in tqdm(range(len(A2))):
    for j in range(len(C)):
        if C.iloc[j,0]==A2.iloc[i,2]:
            C.iloc[j,1]+=1
        elif C.iloc[j,0]==A2.iloc[i,4]:
            C.iloc[j,2]+=1
        else:
            pass
```

```python
for j in range(len(C)):
    W=C.loc[j,'Win']
    L=C.loc[j,'Lose']
    if W+L>0:
        C.loc[j,'Ratio']=W/(W+L)
        
display(C)
```

```python
for i in range(len(B)):
    for j in range(len(C)):
        if B.loc[i,'Team1']==C.loc[j,'Team']:
            B.loc[i,'Team1_Ratio']=C.loc[j,'Ratio']
        elif B.loc[i,'Team2']==C.loc[j,'Team']:
            B.loc[i,'Team2_Ratio']=C.loc[j,'Ratio']

display(B)
```

```python
for i in range(len(B)):
    B.loc[i,'Pred']=0.5+(B.loc[i,'Team1_Ratio']-B.loc[i,'Team2_Ratio'])
    
display(B)
```

```python
B2 = B[['ID','Pred']]
display(B2)
B2.to_csv('submission.csv',index=False)
```
