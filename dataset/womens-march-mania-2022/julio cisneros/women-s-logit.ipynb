```python
# This is my first notebook. Hope it helps
#My approach is simple using Logistic Regression probability.
#I divided the dataset into two parts: First, for predicting results in which we have historic data of both teams competing
#Second, for predicting individual results of each team (as we don't have previous information of both teams competing one vs the other)
# For the first approach, I tried to predict if for the requested game it would result into a win for the first team listed.
#Then for the second,as there was no information about previous matches for both specific teams competing, I compared the probability of winning of one team vs the prob. of winning of the second team in the match
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
#I simplified the information by having a file both with Regular season and NCAA matches, creating a column mentioning who won and lost.

Both= pd.read_csv("../input/bothcompact-onlysubmatch/BothCompact_onlysubmatch.csv")
```

```python
#Then I created dummies for each Team and matches
Trainboth = pd.get_dummies(Both['TeamID'])
Trainmatch = pd.get_dummies(Both['Match'])
Traindumm= Both.join(Trainboth)
TrainR= Traindumm.join(Trainmatch)
TrainR.drop('Match', axis=1, inplace=True)
TrainR.drop('Season', axis=1, inplace=True)
TrainR.drop('DayNum', axis=1, inplace=True)
TrainR.drop('TeamID', axis=1, inplace=True)
TrainR.head()
```

```python
#chose the X & y axis
X_train= TrainR.drop('WIN', axis=1)
y_train = TrainR['WIN']
```

```python
#Then trained a Logistic Regression model
logit       = LogisticRegression(max_iter=200)
model_logit = logit.fit(X_train, y_train)
```

```python
#For the test file I created a quick file dividing the TeamIDs

Test= pd.read_csv("../input/wsamplesubmissionstage2-onlyboth/WSampleSubmissionStage2_onlyboth.csv")
```

```python
#And created dummies for TeamID & Match
Testdum = pd.get_dummies(Test['TeamID'])
Testdummatch = pd.get_dummies(Test['Match'])
Test1= Test.join(Testdum)
Test2= Test1.join(Testdummatch)
Test2.drop('Year', axis=1, inplace=True)
Test2.drop('TeamID', axis=1, inplace=True)
Test2.drop('TeamID2', axis=1, inplace=True)
Test2.drop('Match', axis=1, inplace=True)
Test2.head()
```

```python
#Got then the probability for each Team
prediction = model_logit.predict_proba(Test2)
```

```python
#Used the same previous file for creating predicted values
Submit=pd.read_csv("../input/wsamplesubmissionstage2-onlyboth/WSampleSubmissionStage2_onlyboth.csv")
```

```python
Submit['Pred'] = prediction[:,0]
Submit.head()
```
