The seed benchmark was [described](https://www.kaggle.com/c/march-machine-learning-mania-2014/discussion/6776) by [@jeffsonas](https://www.kaggle.com/jeffsonas) in 2014. It predicts the probability of Team 1 beating Team 2 as:

p(1) = 0.50 + 0.03(Seed(2) - Seed(1)).

This gives a 50%-50% prediction for games between equal seeds, 53%-47% in favour of a team seeded one spot better than their opponents, and 95%-5% for a number 1 seed versus a number 16 seed.

The function is unchanged since its introduction by Sonas in 2014, and hence is independent of any of the test data. Hence, this is a leak-free methodology. 

```python
sub1 = pd.read_csv("../input/ncaaw-2022/stage2_seeds_sample_submission.csv")
sub1.sort_values(by=['ID'], inplace=True)
sub2 = pd.read_csv("../input/womens-march-mania-2022/WDataFiles_Stage2/WSampleSubmissionStage2.csv")
sub2.sort_values(by=['ID'], inplace=True)
prob = 0.5 + 0.03*(sub1['T2_seed'] - sub1['T1_seed'])
prob = np.round(prob, decimals = 2)
sub2['Pred'] = np.clip(prob, 0.05, 0.95)
sub2.to_csv('submission.csv', index=False)
sub2.head()
```
