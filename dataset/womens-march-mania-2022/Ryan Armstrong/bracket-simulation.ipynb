# Run the bracket simulation tool
This notebook makes it very fast and easy to set up a point and click bracket simulation tool using a kaggle submission. The ngrok tunnel that allows this app to work in a Kaggle notebook was inspired by [this great notebook](https://www.kaggle.com/mmotoki/men-s-march-madness-2021-leaderboard-analyzer)

You will need an authentication token from ngrok, which is free, but does require sign up.

Other than that, all you need is a kaggle submission file and the ability to run the cells in this notebook.

Features available in the simulation tool:
* Works for both the men's and women's tournaments
* If submission has multiple years choose any year
* If available it fills in the backet with historical game outcomes (I have done these back to 2016 for the Men's tournament)
* Alternatively, you can simulate a bracket with predicted odds by choosing a random seed or just selecting the favored team.
* Allows you to manually override individual games
* Gives you some summary statistics at the end based on your predictions and the selected outcome, like log loss and accuracy
* Let's you download a summary table of the game outcomes you chose
