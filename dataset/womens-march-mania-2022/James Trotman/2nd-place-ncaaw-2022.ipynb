```python
import pandas as pd
import numpy as np
import os, gc, sys, re, time
import matplotlib.pyplot as plt
from collections import Counter
from itertools import combinations
from pathlib import Path
```

# Read

```python
SEED = 9
YEAR = 2022

INPUT_PATH = Path("../input/womens-march-mania-2022/WDataFiles_Stage2")
CH = "W"

seeds = pd.read_csv(INPUT_PATH / f"{CH}NCAATourneySeeds.csv")
seasons = pd.read_csv(INPUT_PATH / f"{CH}Seasons.csv")
slots = pd.read_csv(INPUT_PATH / f"{CH}NCAATourneySlots2022.csv")
team_names = pd.read_csv(INPUT_PATH / f"{CH}TeamSpellings.csv", encoding="cp1252",
                         index_col="TeamNameSpelling")

five38 = pd.read_csv("fivethirtyeight_ncaa_forecasts.csv")
```

# Prepare

```python
season_query = "Season==" + str(YEAR)
seeds = seeds.query(season_query).copy()
seed2team = seeds.set_index("Seed").TeamID

slot_lookup = dict()
slot_lookup.update(slots.set_index("StrongSeed").Slot.to_dict())
slot_lookup.update(slots.set_index("WeakSeed").Slot.to_dict())

# e.g. W01 = [W01, R1W1, R2W1, R3W1, R4W1, R5WX, R6CH]
def tournament_path(seed):
    key = seed
    l = [seed]
    while key in slot_lookup:
        key = slot_lookup[key]
        l.append(key)
    return l

# e.g. W01 : {R2W1, R4W1, R5WX, W01, R1W1, R3W1, R6CH}
seeds["path"] = seeds.Seed.map(tournament_path).apply(set)
seed2path = seeds.set_index("Seed").path

regions_dict = {v: k for k, v in seasons.query(season_query).iloc[0].items()}
region_code = five38.team_region.str.replace(" ", "").map(regions_dict).str[-1]

idx = five38.team_seed.str.len() == 1
five38.loc[idx, "team_seed"] = "0" + five38.loc[idx, "team_seed"]
five38["seed"] = region_code + five38.team_seed
five38["kg_id"] = five38.team_name.str.lower().map(team_names.TeamID)
five38 = five38.set_index("kg_id")
```

# Baseline

`five38` dataframe is now indexed by Kaggle team IDs

```python
sub = pd.DataFrame(list(combinations(sorted(five38.index), 2)),
                   columns = ["LTeamID", "HTeamID"])
sub.insert(0, "Season", YEAR)

for c in "LH":
    sub[f"{c}Seed"] = sub[f"{c}TeamID"].map(five38.seed)
    sub[f"{c}Rating"] = sub[f"{c}TeamID"].map(five38.team_rating)

# See: https://github.com/jfcross4/kaggle_NCAA/blob/master/stage2_womens.R
# ... for a more elaborate version of this.
# Using this formula alone is a simplification that leads to over-confident predictions.
pred = sub.eval("1./(1+10**((HRating-LRating)*(30.464/400)))")
sub["P1"] = pred
sub["P2"] = pred  # overridden by rd2_win

# The tournament round that two teams meet in is the length of the intersection
# of each team's pathway through the bracket
sub["Round"] = 7 - sub.apply(lambda row: len(seed2path[row.LSeed] & seed2path[row.HSeed]), axis=1)

r1_idx = sub.Round == 1
sub.loc[r1_idx, "P2"] = sub.loc[r1_idx, "LTeamID"].map(five38.rd2_win)

sub.to_csv(f"debug.csv", index=False)
```

# Overrides


```python
P = "Pred"
sub[P] = sub["P2"]
sub["ID"] = (sub["Season"].map(str) + "_" +
             sub["LTeamID"].map(str) + "_" +
             sub["HTeamID"].map(str))
sub = sub.set_index("ID")

threshold, prob_adj, noise_scale = [0.66, 0.14, 1e-05]

logistic_prediction = sub[P]
logistic_confidence = (sub[P] - 0.5).abs() + 0.5
# adjust low confidence predictions
is_confident = logistic_confidence >= threshold
n_pred = sub.shape[0]
b = np.ones(n_pred) * 0.5

np.random.seed(SEED)
# this line adds some insignificant noise
val = np.random.normal(prob_adj, noise_scale, n_pred)
# diverge the predictions 50:50 high/low
val = np.where(np.random.rand(n_pred) < 0.5, val, -val)
```

```python
sub[P] = logistic_prediction.where(is_confident, b + val)
sub[[P]].to_csv(f"sub_a.csv")
```
