```python
INPUT = Path('/kaggle/input/herbarium-2022-fgvc9/')
```

```python
with open(INPUT/'train_metadata.json', 'r') as f:
    data = json.load(f)
```

```python
df_annotations = pd.json_normalize(data['annotations'])
df_images = pd.json_normalize(data['images'])
df_categories = pd.json_normalize(data['categories'])
df_genera = pd.json_normalize(data['genera'])
df_institutions = pd.json_normalize(data['institutions'])
df_distances = pd.json_normalize(data['distances'])
df_license = pd.json_normalize(data['license'])
```
