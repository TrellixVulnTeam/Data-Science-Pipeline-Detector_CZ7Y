```python
import numpy as np
import pandas as pd 
import json
```

```python
TRAIN_IMAGE_FOLDER = "../input/herbarium-2022-fgvc9/train_images/"
TRAIN_JSON = "../input/herbarium-2022-fgvc9/train_metadata.json"
NEW_TRAIN_CSV_PATH = "train.csv"


TEST_IMAGE_FOLDER = "../input/herbarium-2022-fgvc9/test_images/"
TEST_JSON = "../input/herbarium-2022-fgvc9/test_metadata.json"
NEW_TEST_CSV_PATH = "test.csv"

IMAGE_ID = "image_id"
IMAGE_FILE_NAME = "file_name"
IMAGE_TARGET = "institution_id"
IMAGE_FILE_PATH = "image_path"
```

# make train csv 

```python
with open(TRAIN_JSON) as json_file:
    train_data = json.load(json_file)
```

```python
train_annotations = train_data["annotations"]
print("train size=",len(train_annotations))
train_annotations[:3]
```

```python
# train_genus_ids = [train_annotation["genus_id"] for train_annotation in train_annotations]
train_institution_ids = [train_annotation[IMAGE_TARGET] for train_annotation in train_annotations]
# train_category_ids = [train_annotation["category_id"] for train_annotation in train_annotations]
```

```python
train_images = train_data["images"]
train_images[:3]
```

```python
train_image_ids = [train_image[IMAGE_ID] for train_image in train_images]
# train_file_names = [train_image[IMAGE_FILE_NAME] for train_image in train_images]
# train_licenses = [train_image["license"] for train_image in train_images]

train_image_file_paths = [TRAIN_IMAGE_FOLDER + train_image[IMAGE_FILE_NAME] for train_image in train_images]
```

```python
train = pd.DataFrame(data =np.array([train_image_ids ,train_image_file_paths,train_institution_ids]).T, 
                     columns = [IMAGE_ID, IMAGE_FILE_PATH,IMAGE_TARGET,])
train.to_csv(NEW_TRAIN_CSV_PATH, index = False)
train.head()
```

```python
train.tail()
```

# make test data

```python
with open(TEST_JSON) as json_file:
    test_data = json.load(json_file)
    
test_data[:3]
```

```python
test_image_ids = [image[IMAGE_ID] for image in test_data]
test_image_path = [TEST_IMAGE_FOLDER + image[IMAGE_FILE_NAME] for image in test_data ]
```

```python
test = pd.DataFrame(data =np.array([test_image_ids , test_image_path ]).T, 
                    columns = [IMAGE_ID, IMAGE_FILE_PATH])

test.to_csv(NEW_TEST_CSV_PATH, index = False)
test.head()
```
