Due to the large size of the data for this comp, lots of people are trying to reduce memory usage by changing data types of columns, and most solutions I've seen for that use pandas. I found the pandas way painfully slow though, so after some digging around I've found a much faster way with the datatable package. Enjoy!

```python
import datatable as dt
import numpy as np 
import pandas as pd
```

# Change data types using datatable

```python
train_dt = dt.fread('../input/tabular-playground-series-oct-2021/train.csv')
test_dt  = dt.fread('../input/tabular-playground-series-oct-2021/test.csv')
```

```python
%%time
for i,col in enumerate(train_dt):
    if col.type.name == 'float64':
        train_dt[:,i] = dt.as_type(col,'float32')
for i,col in enumerate(test_dt):
    if col.type.name == 'float64':
        test_dt[:,i] = dt.as_type(col,'float32')
train_dt = train_dt.to_pandas()
test_dt = test_dt.to_pandas()
```

# Change data types using pandas

```python
train_pd = dt.fread('../input/tabular-playground-series-oct-2021/train.csv').to_pandas()
test_pd  = dt.fread('../input/tabular-playground-series-oct-2021/test.csv').to_pandas()
```

```python
%%time
for col in train_pd.columns:
    if train_pd[col].dtype == 'float64':
        train_pd[col] = train_pd[col].astype('float32')
        test_pd[col] = train_pd[col].astype('float32')
```
