# Cross Validation Splits

This notebooks generates 5 and 10 folds (stratified) for the dataset.

A new column Fold is added to train.csv. This can be used during training for 5/10 fold cross validation

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import StratifiedKFold
```

```python
train_df = pd.read_csv('../input/tabular-playground-series-oct-2021/train.csv')
```

```python
features = [col for col in train_df.columns if col.startswith('f')]
target = ['target']
```

```python
def generate_n_splits(n=10):
    
    # Get Folds
    skf = StratifiedKFold(n_splits=n,random_state=42,shuffle=True)
    skf.get_n_splits(train_df[features],train_df[target])
    for idx,(train_index,val_index) in enumerate(skf.split(train_df[features],train_df[target])):
        train_df.loc[val_index ,'Fold'] = idx
    
    # Check Folds
    for i in range(n):
        print(f'*****Fold {i}*****')
        print(train_df[train_df['Fold'] == i][target].value_counts(normalize=True))
        print(f'******************')
        
    # Save Folds
    train_df.to_csv(f'Stratified{n}Fold_OCT2021_TPS.csv',index=False)
```

```python
#Generate 5 Folds
generate_n_splits(n=5)
```

```python
#Generate 10 Folds
generate_n_splits(n=10)
```
