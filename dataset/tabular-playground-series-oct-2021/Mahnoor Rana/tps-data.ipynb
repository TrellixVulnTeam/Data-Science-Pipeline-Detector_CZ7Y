```python
x = train.drop('target', axis = 1)
y = train['target']
# splitting data 
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x , y)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 4)

```

```python
# fit the models 

knn.fit(X_train,y_train)
```

```python
# predict the model 

prediction1 = knn.predict_proba(X_test)
```

```python
prediction1
```

```python
# roc_curve
from sklearn.metrics import roc_curve
fpr,tpr,threshold = roc_curve(y_test, prediction1[:,1])
```

```python
# AUC score 
from sklearn.metrics import roc_auc_score
score = roc_auc_score(y_test, prediction1[:,1])
print(score)
```

```python
# plot roc curve 
plt.plot(fpr,tpr,linestyle = '--',color = 'red')
plt.title('roc_curve')
plt.xlabel('fpr')
plt.ylabel('tpr')
plt.show()
```

```python
import lightgbm as lgb
LGB = lgb.LGBMRegressor(random_state = 20,n_estimators = 4399,boosting_type='gbdt')
LGB.fit(X_train,y_train)
```

```python
lgb_pred = LGB.predict(X_test)
lgb_pred
prediction =lgb_pred.sum()
```

```python
#mean square error 
from sklearn.metrics import mean_squared_error
mse = np.sqrt(mean_squared_error(y_test,lgb_pred ))
mse
```

```python
#  submission 
submission = pd.DataFrame({'id': test['id'],
                          'target':prediction})
submission.to_csv('submission.csv',index = False)
```
