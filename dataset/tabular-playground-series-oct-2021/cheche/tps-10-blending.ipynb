```python
subm1 = pd.read_csv('../input/tps-submission/HistGBM_85651.csv')
subm2 = pd.read_csv('../input/tps-submission/submission_85636.csv')
subm3 = pd.read_csv('../input/tps-submission/submission_85637.csv')
subm4 = pd.read_csv('../input/tps-submission/submission_85654.csv')
subm5 = pd.read_csv('../input/tps-submission/submission_85656.csv')
subm6 = pd.read_csv('../input/tps-submission/submission_85658.csv')
```

```python
subm_list = [subm1, subm3, subm4, subm5, subm6]

y_pred = np.zeros((subm1.shape[0],))
for subm in subm_list:
    y_pred += subm.target / 3.9
```

```python
from sklearn.preprocessing import MinMaxScaler

y_pred = y_pred.values.reshape((-1,1))
scaler = MinMaxScaler()
y_pred = scaler.fit_transform(y_pred)
```

```python
submission = pd.read_csv('../input/tabular-playground-series-oct-2021/sample_submission.csv')
submission.target = y_pred
submission.to_csv('submission.csv', index=False)
```
