```python
import numpy as np
import pandas as pd
import os
```

```python
train_set = pd.read_csv("../input/tabular-playground-series-oct-2021/train.csv")
```

```python
train_data = train_set.drop("target", axis=1)
train_label = train_set["target"].copy()
```

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
num_pipeline = Pipeline([
    ('imputer',SimpleImputer(strategy="median")),
    ('std_scaler', StandardScaler()),
])
train_prepared = num_pipeline.fit_transform(train_data)
```

```python
import lightgbm as lgb
lgbm_cls = lgb.LGBMClassifier()
lgbm_cls.fit(train_prepared,train_label)
```

```python
test_set = pd.read_csv("../input/tabular-playground-series-oct-2021/test.csv")
test_prepared = num_pipeline.fit_transform(test_set)
test_pred = lgbm_cls.predict(test_prepared)
```

```python
sub = pd.read_csv('../input/tabular-playground-series-oct-2021/sample_submission.csv')
sub['target'] = list(map(float, test_pred))
sub.to_csv('submission.csv', index=False)
```
