```python
#import packages
import numpy as np
import pandas as pd 
import h2o
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train_pd = pd.read_csv('../input/tabular-playground-series-oct-2021/train.csv')
test_pd = pd.read_csv('../input/tabular-playground-series-oct-2021/test.csv')
```

```python
fig = plt.subplots(figsize = (5,5))
sns.countplot(x='target',data=train_pd)
```

```python
#categorical values

cat_var=[]
for col in train_pd.columns:
    if train_pd[col].dtype=='int64': cat_var.append(col)
train=train_pd[0:5000]
test=test_pd[0:5000]
```

```python
# Boxplot of categorical variables

plt.figure(figsize=(30, 30))
sns.set(style="darkgrid")
i=1
for col in cat_var:
    plt.subplot(12,4,i)
    sns.countplot(data=train, x=col)
    i=i+1
plt.show()
```

```python
plt.figure(figsize=(20 , 20))
corr = train.corr()
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
sns.heatmap(corr, square=True, mask = mask, cmap='coolwarm_r', annot_kws={'size':20})
```

Since, it is tough to look at the values , let plot the correlation matrix only for those columns where correaltion is greater than 0.5.

```python
m = (corr.mask(np.eye(len(corr), dtype=bool)).abs() > 0.5).any()
raw = corr.loc[m, m]
raw
```

```python
#histogram of numeric values.
num_col= [col for col in train_pd.columns if (col not in cat_var) ]
num_col1=num_col[:49]
plt.figure(figsize=(100, 100))
i=1
for col in num_col1:
    plt.subplot(10,5,i)
    sns.histplot(data=train, x=col)
    plt.title(col+" distribution")
    i=i+1
plt.show()
```
