```python
from pandas import read_csv
import sweetviz as sv
```

```python
# Load dataset
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
        
filename_training          = '/kaggle/input/tabular-playground-series-oct-2021/train.csv'
filename_test              = '/kaggle/input/tabular-playground-series-oct-2021/test.csv'
filename_sample_solution   = '/kaggle/input/tabular-playground-series-oct-2021/sample_submission.csv'
dataset_sample_solution    = read_csv(filename_sample_solution)
dataset_training           = read_csv(filename_training)
dataset_test               = read_csv(filename_test)
```

```python
advert_report = sv.analyze([dataset_training, 'target'], pairwise_analysis='off')
```
