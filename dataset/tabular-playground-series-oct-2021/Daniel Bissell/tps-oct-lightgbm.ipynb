```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Read the data
X = pd.read_csv('/kaggle/input/tabular-playground-series-oct-2021/train.csv', index_col='id')
X_test_full = pd.read_csv('/kaggle/input/tabular-playground-series-oct-2021/test.csv', index_col='id')
# Remove rows with missing target, separate target from predictors
y = X.target              
X.drop(['target'], axis=1, inplace=True)
```

```python
X_train, X_valid, y_train, y_valid = \
    train_test_split(X, y, stratify=y, train_size=0.75)
```

```python
from lightgbm import LGBMRegressor
model = LGBMRegressor(random_state = 0, metric = "auc")
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import roc_auc_score

print("training dataset auc is {0}".format(roc_auc_score(y_train, model.predict(X_train))))
print("validation dataset auc is {0}".format(roc_auc_score(y_valid, model.predict(X_valid))))
```

```python
X_testy = pd.read_csv('/kaggle/input/tabular-playground-series-oct-2021/test.csv', index_col='id')
```

```python
Yhat = model.predict(X_testy)
```

```python
output = pd.DataFrame({'id': X_testy.index,
                       'target': Yhat})
output.to_csv('submission.csv', index=False)
```
