```python
import numpy as np
import pandas as pd
```

```python
lgb_categ = pd.read_csv("../input/kaggleoctplayground/submission_lgb_categ.csv")
lgb_num = pd.read_csv("../input/kaggleoctplayground/submission_lgb_numerical.csv")
xgb_categ = pd.read_csv("../input/kaggleoctplayground/submission_xgboost_categorial.csv")
xgb_num = pd.read_csv("../input/kaggleoctplayground/submission_xgboost_numerical.csv")
```

```python
submission = pd.DataFrame(lgb_categ.id)
submission.head()
```

```python
predictions = (0.7 * xgb_num['target'].values + 0.5 * xgb_categ['target'].values + 
               lgb_num['target'].values + lgb_categ['target'].values) / 4
```

```python
submission['target'] = predictions
submission.to_csv('submission.csv', index=False)
submission.head()
```
