```python
df1=pd.read_csv('../input/temporary-outputs/xgb_clf_predictions.csv')
df2=pd.read_csv('../input/temporary-outputs/lgbm_clf_predictions.csv')
df3=pd.read_csv('../input/cat-boost-output/New folder/catb_clf_predictions.csv')
```

```python
df3.head()
```

```python
df4=df1.merge(df2,on='id')
df4=df4.merge(df3,on='id')
```

```python
df4.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
lr=LogisticRegression()
train_df=df4[df4.kfold_x!=4]
test_df=df4[df4.kfold_x==4]

x_train=train_df[['xgbclf_pred','lgbmclf_pred','catbclf_pred']]
x_test=test_df[['xgbclf_pred','lgbmclf_pred','catbclf_pred']]
y_train=train_df.target_x
y_test=test_df.target_x
```

```python
import sklearn
import joblib
```

```python
lr.fit(x_train,y_train)
y_pred=lr.predict_proba(x_test)[:,1]
roc=sklearn.metrics.roc_auc_score(y_test,y_pred)
print(roc)
```
