# Import libraries and Load dataset

```python
import numpy as np
import pandas as pd
import tensorflow as tf

import autokeras as ak
```

```python
train = pd.read_csv("../input/tabular-playground-series-oct-2021/train.csv",nrows=100000)
```

### As you know, TPS-Oct has so many data. So, I will use only 100000 rows on train data

```python
train = train.drop('id',axis=1)
y_train = train['target']
X_train = train.drop('target',axis=1)

X_train = X_train.to_numpy()
y_train = y_train.to_numpy()
```

### Drop the id column and split the train dataset to X_train, y_train data

# Create and Train Model !

```python
clf = ak.StructuredDataClassifier(overwrite=True, max_trials=5)
clf.fit(X_train, y_train, epochs=10, validation_split=0.3)
```

### AutoKeras make model autoly, So you just set Classifier or Regression, (you can make model for Image and Text data)
### Just waiting! Then, Autokeras make auto model! It's done

```python
model = clf.export_model()
model.summary()
```

### you can check the best model through this code.
### after the predict process is same with just using other keras model.

# Update ---------------------------------------------
### I will use best model, Train all data !

```python
train = pd.read_csv("../input/tabular-playground-series-oct-2021/train.csv")
test = pd.read_csv("../input/tabular-playground-series-oct-2021/test.csv")
train = train.drop('id',axis=1)
y_train = train['target']
X_train = train.drop('target',axis=1)

X_train = X_train.to_numpy()
y_train = y_train.to_numpy()

test = test.drop('id',axis=1)
test = test.to_numpy()

model._layers.pop()
model._layers.pop()

tmp = tf.keras.Sequential(model)
tmp.add(tf.keras.layers.Activation('tanh'))
tmp.compile(optimizer='adam', loss='mse')

tmp.fit(X_train, y_train, epochs=5, validation_split=0.3)
```

```python
pred = tmp.predict(test)
sample = pd.read_csv("../input/tabular-playground-series-oct-2021/sample_submission.csv")
sample['target'] = pred
sample.to_csv("test.csv",index=None)
```

### The AutoKeras's performance is not good in Structed Dataset Like TPS-Oct
