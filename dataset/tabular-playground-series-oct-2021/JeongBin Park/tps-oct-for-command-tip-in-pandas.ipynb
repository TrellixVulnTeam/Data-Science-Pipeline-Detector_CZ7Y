## Data Load

```python
import pandas as pd
import time
my_data = pd.read_csv('../input/tabular-playground-series-oct-2021/train.csv',usecols = ['id','f0'])
my_data['accumulate_f0'] = 0
print("data's length : ", len(my_data))
```

### This TPS - Oct data has 1,000,000 rows !
### Just think that make accumulate value. I will compare four ways

## 1. Using len()

```python
start_time = time.time()
tmp_arr = [0]
for i in range(1, len(my_data)):
    row_prev, row = my_data.loc[i-1,'f0'], my_data.loc[i,'f0']
    tmp_arr.append(row + row_prev)
my_data.loc[:,'accumulate_f0'] = tmp_arr
print("Time : ", time.time() - start_time)
```

### I think many people using this way in many situation in python include pandas. 
### But, it take the longest time badly...

## 2. iterrows()

```python
start_time = time.time()
tmp_arr = []
prev = 0
for i, row in my_data.iterrows():
    prev = row['f0'] + prev
    tmp_arr.append(prev)
my_data.loc[:,'accumulate_f0'] = tmp_arr
print("Time : ", time.time() - start_time)
```

### Do you know that pandas has iterrows() function?
### I didn't know that. 
### you just think [Pandas] : iterrows , [Python] : enumerate
### It's faster than using len(), But still slow

## 3. itertuples()

```python
start_time = time.time()
tmp_arr = []
prev = 0
for tuple_data in my_data.itertuples():
    prev = tuple_data.f0 + prev
    tmp_arr.append(prev)
my_data.loc[:,'accumulate_f0'] = tmp_arr
print("Time : ", time.time() - start_time)
```

### Pandas has itertuples function. 
### It's special dataformat in pandas. So very fast ! 
### Look I use tuple_data.f0 not tuple_data['f0']
### tuple_data['f0'] is not works in using itertuples()

## 4. Using pandas function

```python
start_time = time.time()
my_data.loc[:,'accumulate_f0'] = my_data['f0'].cumsum()
print("Time : ", time.time() - start_time)
```

### If you have a function that Pandas supports, just use it. That's the fastest.
