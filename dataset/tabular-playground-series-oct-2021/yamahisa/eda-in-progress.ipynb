```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")

pd.set_option('display.max_rows', 300)
pd.set_option('display.max_columns', 300)

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

**First, check what the difference is between the train data and the test data.**

```python
train = pd.read_csv('../input/tabular-playground-series-oct-2021/train.csv')
test = pd.read_csv('../input/tabular-playground-series-oct-2021/test.csv')
```

```python
eda_sa = pd.DataFrame({"train_mean" :train.iloc[:,1:-1].mean(),
                         "test_mean":test.iloc[:,1:].mean(),
                         "train_std":train.iloc[:,1:-1].std(),
                         "test_std":test.iloc[:,1:].std(),
                         "train_skew":train.iloc[:,1:-1].skew(),
                         "test_skew":test.iloc[:,1:].skew(),
                         "train_kur":train.iloc[:,1:-1].kurt(),
                         "test_kur":test.iloc[:,1:].kurt(),
                         "train_max":train.iloc[:,1:-1].max(),
                         "test_max":test.iloc[:,1:].max(),
                         "train_min":train.iloc[:,1:-1].min(),
                         "test_min":test.iloc[:,1:].min()
                        })
```

```python
fig = plt.figure(figsize = [14,8])
plt.plot(eda_sa["train_mean"], alpha = 0.5, label = "train")
plt.plot(eda_sa["test_mean"], alpha = 0.5, label = "test")
plt.legend(loc = "upper right")
plt.show()
```

```python
fig = plt.figure(figsize = [14,8])
plt.plot(eda_sa["train_std"], alpha = 0.5, label = "train")
plt.plot(eda_sa["test_std"], alpha = 0.5, label = "test")
plt.legend(loc = "upper right")
plt.show()
```

```python
fig = plt.figure(figsize = [14,8])
plt.plot(eda_sa["train_skew"], alpha = 0.5, label = "train")
plt.plot(eda_sa["test_skew"], alpha = 0.5, label = "test")
plt.legend(loc = "upper right")
plt.show()
```

```python
fig = plt.figure(figsize = [14,8])
plt.plot(eda_sa["train_kur"], alpha = 0.5, label = "train")
plt.plot(eda_sa["test_kur"], alpha = 0.5, label = "test")
plt.legend(loc = "upper right")
plt.show()
```

```python
fig = plt.figure(figsize = [14,8])
plt.plot(eda_sa["train_max"], alpha = 0.5, label = "train")
plt.plot(eda_sa["test_max"], alpha = 0.5, label = "test")
plt.legend(loc = "upper right")
plt.show()
```

```python
fig = plt.figure(figsize = [14,8])
plt.plot(eda_sa["train_min"], alpha = 0.5, label = "train")
plt.plot(eda_sa["test_min"], alpha = 0.5, label = "test")
plt.legend(loc = "upper right")
plt.show()
```

There is a difference between the maximum value and the minimum value, There seems to be almost no difference in mean, standard deviation, skewness, and kurtosis.

There seems to be no big difference between train data and test data.

**Next, perform an analysis of training data
**

```python
train.target.value_counts()
```

The target data of the training data is binary data and contains almost the same number.

```python
train.dtypes
```

```python
train_target0 = train[train["target"] == 0]
train_target1 = train[train["target"] == 1]
train_target0.drop(["target", "id"], axis = 1, inplace = True)
train_target1.drop(["target", "id"], axis = 1, inplace = True)
```

```python
eda_tsa = pd.DataFrame({"train_target0_mean" :train_target0.mean(),
                        "train_target1_mean"  :train_target1.mean(),
                        "train_target0_skew":train_target0.skew(),
                        "train_target1_skew":train_target1.skew(),
                        "train_target0_kur":train_target0.kurt(),
                        "train_target1_kur":train_target1.kurt(),
                        "train_mean":train.iloc[:, 1:-1].mean(),
                        "train_skew":train.iloc[:, 1:-1].skew(),
                        "train_kur":train.iloc[:, 1:-1].kurt(),                        
                        })
eda_tsa.head(3)
```

```python
fig = plt.figure(figsize = [22,8])
plt.plot(eda_tsa["train_target0_mean"], alpha = 0.5, label = "train_target0_mean")
plt.plot(eda_tsa["train_target1_mean"], alpha = 0.5, label = "train_target1_mean")
plt.plot(eda_tsa["train_mean"], alpha = 0.2, label = "train_mean")
plt.legend(loc = "upper right")
plt.show()
```

```python
fig = plt.figure(figsize = [22,8])
plt.plot(eda_tsa["train_target0_skew"], alpha = 0.5, label = "train_target0_skew")
plt.plot(eda_tsa["train_target1_skew"], alpha = 0.5, label = "train_target1_skew")
plt.plot(eda_tsa["train_skew"], alpha = 0.2, label = "train_skew")
plt.legend(loc = "upper right")
plt.show()
```

```python
fig = plt.figure(figsize = [22,8])
plt.plot(eda_tsa["train_target0_kur"], alpha = 0.5, label = "train_target0_kurt")
plt.plot(eda_tsa["train_target1_kur"], alpha = 0.5, label = "train_target1_kurt")
plt.plot(eda_tsa["train_kur"], alpha = 0.2, label = "train_kurt")
plt.legend(loc = "upper right")
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

forest = RandomForestClassifier(criterion = "gini",
                                n_estimators = 25,
                                random_state = 1,
                                n_jobs = -1)

X = train.iloc[:,1:-1]
y = train.iloc[:, -1]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0, stratify = y)
forest.fit(X_train, y_train)

print("training dataset score : {:.3f} / test dataset score:{:.3f}".format(forest.score(X_train, y_train), forest.score(X_test, y_test)))
```
