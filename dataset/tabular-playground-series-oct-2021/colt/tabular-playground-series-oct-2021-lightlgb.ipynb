```python
#Data Read
data_X=pd.read_csv("../input/tabular-playground-series-oct-2021/test.csv")
data_y=pd.read_csv("../input/tabular-playground-series-oct-2021/train.csv")

print(data_X.head())
print(data_y.head())
```

```python
#Check Null
def fillCheck(df):
        n_val=df.isnull().sum()
        pre=100*df.isnull().sum()/len(df)
        tdble=pd.concat([n_val,pre],axis=1)
        ren_colum=tdble.rename(columns={0:"Null Persent",1:"%"})
        return ren_colum
    
fillCheck(data_X) 
fillCheck(data_y) 
```

```python
#Data_Split
df_train,df_val=train_test_split(data_y,test_size=0.5)

X_train=df_train.drop("target",axis=1)
y_train=df_train["target"]

X_val=df_val.drop("target",axis=1)
y_val=df_val["target"]
```

```python
#LGB Modeling
params = {
    'random_state':100,
    'verbose':0,
    #Learning model indicators
    'metrics' : "auc"
}


train_d=lgb.Dataset(X_train,y_train)
val_d=lgb.Dataset(X_val,y_val)

model=lgb.train(params,train_d,valid_sets=val_d,num_boost_round=100,early_stopping_rounds=100)
```

```python
#fitting
pred=model.predict(data_X)
```

```python
#submit
PassengerId=data_X["id"]
print(PassengerId.shape)
print(PassengerId.head())
Predictions=pd.DataFrame(data=pred,index=PassengerId,columns=['Target'])
Predictions.to_csv("submission.csv")
```
