<h1>Imports</h1>

<h2>Packages</h2>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

<h2>Training Data</h2>

```python
train_data = pd.read_csv('../input/tabular-playground-series-oct-2021/train.csv')
train_data.head()
```

<h1>Create Sample of Dataset</h1>

<h2>Training Data Sample</h2>

```python
train = train_data.sample(n=10000, random_state=42)
train.head()
```

<h2>Ensure Sample is Representative of Full Data (maybe run this a few times)</h2>

```python
import random

rand_feat = random.randint(0, 284)
print('Feature: f' + str(rand_feat))
sns.kdeplot(data=train_data, x=f'f{rand_feat}', color='blue', fill=True)
sns.kdeplot(data=train, x=f'f{rand_feat}', color='red', fill=True)
```

<h1>Sample Correlation</h1>

<h2>Display all correlation results</h2>

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

<h2>Generate Array of Features Labels (for correlation table formatting)</h2>

```python
temp = 0
subset_array = []

while (temp < 285):
    subset_array.append(f'f{temp}')
    temp += 1

print(subset_array)
```

<h2>Correlation of Training Sample Data</h2>

```python
corr = train.corr()

corr_df = pd.DataFrame(data=corr['target'])
corr_df = corr_df[1:287]
corr_df = corr_df.transpose()

corr_df.style.background_gradient(subset=subset_array, cmap='coolwarm', vmin=-1.0, vmax=1.0)
```

<h2>Sort by Value for Convenience</h2>

```python
corr_df = corr_df.transpose()
corr_df = corr_df.sort_values('target')
corr_df = pd.DataFrame(corr_df)
corr_df = corr_df.transpose()
corr_df.style.background_gradient(subset=subset_array, cmap='coolwarm', vmin=-1.0, vmax=1.0)
```

<h1>Correlation of Full Training Data</h1>

<h2>Correlation of Features to Target - Full "train.csv"</h2>

```python
corr = train_data.corr()

corr_df = pd.DataFrame(data=corr['target'])
corr_df = corr_df[1:287]
corr_df = corr_df.transpose()

corr_df.style.background_gradient(subset=subset_array, cmap='coolwarm', vmin=-1.0, vmax=1.0)
```

<h2>Sorted</h2>

```python
corr_df = corr_df.transpose()
corr_df = corr_df.sort_values('target')
corr_df = pd.DataFrame(corr_df)
corr_df = corr_df.transpose()
corr_df.style.background_gradient(subset=subset_array, cmap='coolwarm', vmin=-1.0, vmax=1.0)
```
