```python
#import packages
import numpy as np
import pandas as pd 
#import matplotlib as mpl
import h2o
from h2o.automl import H2OAutoML
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
h2o.init()
```

```python
train = h2o.import_file('../input/tabular-playground-series-oct-2021/train.csv')
test = h2o.import_file('../input/tabular-playground-series-oct-2021/test.csv')
```

```python
x = train.columns
y = 'target'

x.remove(y)
x.remove('id') #removing id  
```

```python
aml_y = H2OAutoML(max_runtime_secs=60*20, seed=1)
aml_y.train(x=x, y=y, training_frame=train)
```

```python
preds_y = aml_y.predict(test)
```

```python
submission = pd.read_csv("../input/tabular-playground-series-oct-2021/sample_submission.csv")
submission.target = preds_y.as_data_frame().predict
submission.to_csv('submission.csv', index=False)
```
