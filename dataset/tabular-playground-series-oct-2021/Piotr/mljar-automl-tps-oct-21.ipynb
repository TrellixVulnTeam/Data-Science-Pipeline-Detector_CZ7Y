```python
import numpy as np
import pandas as pd
from supervised.automl import AutoML # mljar-supervised
```

```python
train = pd.read_csv("../input/tabular-playground-series-oct-2021/train.csv")
test = pd.read_csv("../input/tabular-playground-series-oct-2021/test.csv")
```

```python
train.head()
```

```python
x_cols = train.columns[1:-1].tolist()
y_col = train.columns[-1]
```

```python
automl = AutoML(
    mode="Compete", 
    total_time_limit=4*3600, # 4 hours
    eval_metric='auc'
)
automl.fit(train[x_cols], train[y_col])
```

```python
preds = automl.predict_proba(test)
```

```python
sub = pd.DataFrame({"id": test.id, "target": preds[:,1]})
sub.to_csv("./1_sub.csv", index=False)
```
