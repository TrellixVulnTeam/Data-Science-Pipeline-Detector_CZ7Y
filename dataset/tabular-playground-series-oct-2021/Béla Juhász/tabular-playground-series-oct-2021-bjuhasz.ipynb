```python
# Source: https://docs.fast.ai/tutorial.tabular

path = '/kaggle/input/tabular-playground-series-oct-2021'
csv_path = f'{path}/train.csv'

training_data = pd.read_csv(csv_path, nrows=10_000)
# training_data.drop('id', axis=1, inplace=True)
training_data.set_index('id', inplace=True)
# training_data.head()

# A target az vagy 0, vagy 1.
# training_data['target'].unique()
```

```python
from fastai import *
from fastai.tabular.all import *

training_data = TabularDataLoaders.from_df(training_data, path=path, y_names='target', y_block=CategoryBlock)
```

```python
from fastai import *

learn = tabular_learner(training_data, metrics=accuracy)
learn.fit_one_cycle(1)
```

```python
csv_path = f'{path}/test.csv'
test_data = pd.read_csv(csv_path, nrows=10_000)
test_data.columns
# test_data.drop('id', axis=1, inplace=True)
test_data.set_index('id', inplace=True)
# test_data.head()
```

```python
# test_data.head()
# learn.predict(test_data)
learn.predict(test_data.iloc[0])
# create submission file to submit in Kaggle competition
# submission = pd.DataFrame({'PassengerId': test_id, 'Survived': labels})
# submission.to_csv('submission.csv', index=False)
# submission.head()

dl = learn.dls.test_dl(test_data)
preds, a = learn.get_preds(dl=dl) # get prediction
print(preds)
print(preds.shape)
```
