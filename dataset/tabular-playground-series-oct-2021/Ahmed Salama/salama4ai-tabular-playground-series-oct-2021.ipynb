```python
from sklearn.pipeline import make_pipeline
from sklearn.feature_selection import VarianceThreshold
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
from xgboost import XGBClassifier
params = {
    'max_depth': 6,
    'n_estimators': 17500,
    'learning_rate': 0.021065232180933688,
    'subsample': 0.9,
    'colsample_bytree': 0.2,
    'colsample_bylevel': 0.30000000000000004,
    'min_child_weight': 292.65754478395786,
    'reg_lambda': 6.287542520595498,
    'reg_alpha': 142.19347996952186,
    'gamma': 0.020280285370407516,
    'booster': 'gbtree',
    'eval_metric': 'auc',
    'tree_method': 'gpu_hist',
    'predictor': 'gpu_predictor',
    'use_label_encoder': False
}
```

```python
thr_drop = VarianceThreshold(0.17)
sc = StandardScaler()
clf = RandomForestClassifier(n_jobs=-1)
pipe = make_pipeline(sc, thr_drop, clf)
```

```python
pipe.fit(x, y)
```

```python
test_df = pd.read_csv(test_csv, index_col=index)
ypred = pipe.predict(test_df)
```

```python
salama4ai_submission = pd.DataFrame({index: pd.read_csv(test_csv)[index], label: ypred})
salama4ai_submission.to_csv("submission.csv", index=False)
```
