```python
import pandas as pd
from lightautoml.automl.presets.tabular_presets import TabularAutoML
from lightautoml.tasks import Task
```

```python
train_data = pd.read_csv('../input/tabular-playground-series-oct-2021/train.csv')
test_data = pd.read_csv('../input/tabular-playground-series-oct-2021/test.csv')
sub = pd.read_csv('../input/tabular-playground-series-oct-2021/sample_submission.csv')
train_data.shape, test_data.shape, sub.shape
```

```python
task = Task('binary')
automl = TabularAutoML(task = task, timeout = 8 * 3600, cpu_limit = 4, 
                       general_params = {'use_algos': [['cb_tuned']]}, 
                       selection_params = {'mode': 0},
                       tuning_params = {'max_tuning_time': 1800})
oof_pred = automl.fit_predict(train_data, roles = {'target': 'target', 'drop': ['id']})
sub['target'] = automl.predict(test_data).data[:, 0]
```

```python
sub.to_csv('submission.csv', index = False)
```
