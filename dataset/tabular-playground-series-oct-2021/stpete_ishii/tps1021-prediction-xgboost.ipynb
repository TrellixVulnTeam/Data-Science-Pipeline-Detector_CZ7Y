# XGBoost Prediction with Optuna tuning
* doc: https://github.com/optuna/optuna

```python
import xgboost as xgb
import numpy as np
import pandas as pd
import random
import optuna
from sklearn.model_selection import KFold, train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
```

```python
train0 = pd.read_csv("../input/tabular-playground-series-oct-2021/train.csv")
test = pd.read_csv("../input/tabular-playground-series-oct-2021/test.csv")
```

```python
train0.head()
```

```python
test.head()
```

```python
n=len(train0)
print(n)
N=list(range(n))
random.seed(2021)
random.shuffle(N)
```

```python
train=train0.iloc[N[0:10000],:]
```

```python
target = train['target']
data = train.drop(['target','id'],axis=1)
test = test.drop('id',axis=1)
```

```python
print(len(data))
print(len(train))
print(len(test))
```

```python
columns=data.columns.to_list()
print(columns)
```

```python
def objective(trial,data=data,target=target):
    
    train_x, test_x, train_y, test_y = train_test_split(data, target, test_size=0.2,random_state=42)
    param = {
        'objective': trial.suggest_categorical('objective',['reg:logistic','reg:tweedie']), 
        'tree_method': trial.suggest_categorical('tree_method',['gpu_hist']),  # 'gpu_hist','hist'
        'lambda': trial.suggest_loguniform('lambda',1e-3,10.0),
        'alpha': trial.suggest_loguniform('alpha',1e-3,10.0),
        'colsample_bytree': trial.suggest_loguniform('colsample_bytree',1e-3,1.0),
        'subsample': trial.suggest_loguniform('subsample', 1e-3,1.0),
        'n_estimators': trial.suggest_int('n_estimators', 2,200),
        'max_depth': trial.suggest_int('max_depth', 2,10),
        'learning_rate': trial.suggest_loguniform('learning_rate',1e-1,1.0),
        'random_state': trial.suggest_int('random_state', 2,2021),
        'min_child_weight': trial.suggest_int('min_child_weight',1,300),
        'use_label_encoder': trial.suggest_categorical('use_label_encoder',[False])
    }
    model = xgb.XGBClassifier(**param)      
    model.fit(train_x,train_y,eval_set=[(test_x,test_y)],early_stopping_rounds=100,verbose=False)
    preds = model.predict(test_x)
    rmse = mean_squared_error(test_y, preds,squared=False)
    
    return rmse
```

```python
study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=8)
print('Number of finished trials:', len(study.trials))
print('Best trial:', study.best_trial.params)
```

```python
# shows the scores from all trials
optuna.visualization.plot_optimization_history(study)
```

```python
# Visualize parameter importances.
optuna.visualization.plot_param_importances(study)
```

```python
# Visualize empirical distribution function
optuna.visualization.plot_edf(study)
```

```python
Best_trial=study.best_trial.params
print(Best_trial)
```

```python
sample = pd.read_csv("../input/tabular-playground-series-oct-2021/sample_submission.csv")
sample
```

```python
preds = np.zeros((sample.shape[0]))
kf = KFold(n_splits=5,random_state=48,shuffle=True)
for trn_idx, test_idx in kf.split(train[columns],target):
    X_tr,X_val=train[columns].iloc[trn_idx],train[columns].iloc[test_idx]
    y_tr,y_val=target.iloc[trn_idx],target.iloc[test_idx]
    model = xgb.XGBClassifier(**Best_trial)
    model.fit(X_tr,y_tr,eval_set=[(X_val,y_val)],early_stopping_rounds=100,verbose=False)
    preds+=model.predict(test[columns])/kf.n_splits   ###### predict_proba
    rmse=mean_squared_error(y_val, model.predict(X_val),squared=False)
    print(rmse)
```
