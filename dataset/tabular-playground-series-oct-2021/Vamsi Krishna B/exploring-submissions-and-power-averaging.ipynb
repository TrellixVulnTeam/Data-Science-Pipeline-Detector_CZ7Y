We are going to use Power Averaging technique,which is a form of ensembling to optimize for the AUC metric.
Here's a detailed post on it - https://medium.com/data-design/reaching-the-depths-of-power-geometric-ensembling-when-targeting-the-auc-metric-2f356ea3250e. 

It works best on highly correlated models.The high positive probabilities of the positively confident models remain positively confident. The non-high probabilities of any models are knocked down. 

The formula is **Final Submission = (Submission1^Power + Submission2^Power + Submission3^Power + Submission4^Power) / 4**

Go through this great notebook https://www.kaggle.com/edrickkesuma/in-depth-power-averaging-0-81848 for better understanding and implementation of power averaging.

```python
# Power is arbitrary - refer to blog post for more info to get a better power
ensemble = b1.copy()
ensemble.loc[:,'target'] = (b1**4 + b2**4 + b3**4 +b4**4+b5**4+b6**4)/6
```
