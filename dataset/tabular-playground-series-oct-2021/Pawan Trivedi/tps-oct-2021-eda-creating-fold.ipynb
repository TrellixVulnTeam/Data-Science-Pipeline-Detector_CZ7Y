```python
# Missing Values

def missing_values(data):
    missing_values =  pd.DataFrame(
    {
       'missing_values':data.isnull().sum(),
       'percentage':data.isnull().sum()*100/data.shape[0]
    })
    missing_values.sort_values(by='missing_values', ascending=False)
    return missing_values.T
```

üìå Creating **folds** and saving data back to csv file. 

```python
train_data['kfold'] = -1

kf = KFold(n_splits=5, shuffle=True, random_state=43)
for fold, (train_indicies, validation_indicies) in enumerate(kf.split(X=train_data)):
    train_data.loc[validation_indicies, 'kfold'] = fold
```

```python
train_data.to_csv('train_fold.csv', index=False)
```
