# Lets Load the Training File

```python
train_df = pd.read_csv('../input/hpa-single-cell-image-classification/train.csv')
train_df.head(10)
```

As we can see, train.csv contains image id and label column.

# Unique Labels

```python
all_labels = train_df.Label.unique().tolist()
```

```python
all_labels[:10]
```

Labels are separated by | symbol

```python
all_labels = '|'.join(all_labels)

all_labels = all_labels.split('|')
```

```python
labels = [str(i) for i in range(19)]

unique_counts = {}
for lbl in labels:
    unique_counts[lbl] = len(train_df[train_df.Label == lbl])

full_counts = {}
for lbl in labels:
    count = 0
    for row_label in train_df['Label']:
        if lbl in row_label.split('|'): count += 1
    full_counts[lbl] = count

df_lbls = pd.DataFrame(index=labels)
df_lbls['full_count'] = full_counts.values()
df_lbls['uniq_count'] = unique_counts.values()
df_lbls.plot.bar(figsize=(16,8))
```
