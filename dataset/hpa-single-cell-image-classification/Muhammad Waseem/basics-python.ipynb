Loop and Functions
A function is a block of code which only runs when it is called.
Data can be passed into functions.
A function can also return a value

```python
df = [2,6,3,7,8,3,1,5,4,7,3,8,2,7]
less_than_5 = 0
greater_than_5 = 0
for i in df:
    if i>5:
        greater_than_5 = greater_than_5 + i
    else:
        less_than_5 = less_than_5 + i
print(greater_than_5,  less_than_5)
```

```python
lst = [2,6,3,7,8,3,1,5,4,7,3,8,2,7]
less_than_5 = 0
greater_than_5 = 0
for i in lst:
    if i>5:
        greater_than_5 = greater_than_5 + i
    else:
        less_than_5 = less_than_5 + i
print(greater_than_5,  less_than_5)
```
