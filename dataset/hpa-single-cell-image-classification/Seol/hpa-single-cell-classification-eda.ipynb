```python
def GetImage(csv, index):
    train_dir = '../input/hpa-single-cell-image-classification/train'
    for i in range(index):
        image_b = train_dir + '/' + csv.iloc[i,0] + '_blue.png'
        image_g = train_dir + '/' + csv.iloc[i,0] + '_green.png'
        image_r = train_dir + '/' + csv.iloc[i,0] + '_red.png'
        image_y = train_dir + '/' + csv.iloc[i,0] + '_yellow.png'
        
        img_b = load_img(image_b)
        img_g = load_img(image_g)
        img_r = load_img(image_r)
        img_y = load_img(image_y)
        
        print()
        fig=plt.figure(figsize=(20, 20))
        cols = 4
        rows = 1
        
        img = [img_b, img_g, img_r, img_y]
        
        for i in range(0, cols):
            fig.add_subplot(rows, cols, i+1)
            plt.imshow(img[i])
            
        plt.show()
```

```python
GetImage(train_csv, 3)
```
