```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from colorama import Fore, Style
import cv2

import plotly.express as px
import plotly.graph_objs as go
import plotly.figure_factory as ff
```

```python
def cout(string: str, color: str, end='\n') -> str:
    """
    Prints a string in the required color
    """
    print(color+string+Style.RESET_ALL, end=end)
```

```python
TRAIN_DIR = "../input/hpa-single-cell-image-classification/train"
TEST_DIR = "../input/hpa-single-cell-image-classification/test"
```

```python
data = pd.read_csv("../input/hpa-single-cell-image-classification/train.csv")
data.sample(10)
```

## Top-10 Most Occurring Labels in the Training Dataset

Let's start by looking at the Top-10 most occurring labels in the training dataset.

```python
top_10_keys = list(dict(data['Label'].value_counts()).keys())[:10]
top_10_values = list(dict(data['Label'].value_counts()).values())[:10]

cout("Top-10 Labels in the Dataset:\n", color=Fore.BLUE)
for i, (x, y) in enumerate(zip(top_10_keys, top_10_values)):
    cout(f"Rank: {i+1}", color=Fore.CYAN, end=' \t')
    cout("Label: ", color=Fore.GREEN, end=' ')
    cout(f"{x} ", color=Fore.RED, end='\t')
    cout("Values: ", color=Fore.GREEN, end=' ')
    cout(f"{y} ", color=Fore.RED, end='\n')
```

## Top-10 Least Occurring Labels in the Training Dataset

Now let us at the Top-10 least occuring labels in the training dataset.

```python
top_10_keys = list(dict(data['Label'].value_counts()).keys())[-10:]
top_10_values = list(dict(data['Label'].value_counts()).values())[-10:]

cout("Top-10 (least occurring) Labels in the Dataset:\n", color=Fore.BLUE)
for i, (x, y) in enumerate(zip(top_10_keys, top_10_values)):
    cout("Label: ", color=Fore.GREEN, end=' ')
    cout(f"{x} ", color=Fore.RED, end='\t')
    cout("Values: ", color=Fore.GREEN, end=' ')
    cout(f"{y} ", color=Fore.RED, end='\n')
```

## Are IDs repeated?

No they are not. But we do have 4 Different Color Images for every single ID.

These 4 different colors (`Red`, `Blue`, `Yellow`, `Green`) are 4 filters for specific ROIs.

## Visualizing Images of a Few Top-Categories

Let's now visualize the images of a few top categories and see what we find.

### Category: 0

```python
# Get all 4 color grade images for a single id
category_0 = data[data['Label'] == '0']['ID']
rdm_cat0_img = np.random.choice(category_0.tolist())
all_imgs_current = []
for x in os.listdir(TRAIN_DIR):
    if rdm_cat0_img in x:
        all_imgs_current.append(x)
```

#### Microtube Channels (Red)

```python
current_red = None
for x in all_imgs_current:
    if "red" in x:
        current_red = x
current_red = cv2.imread(os.path.join(TRAIN_DIR, current_red), 0)
fig = px.imshow(current_red, title='Category: 0, [Microtube Channels (Red)]')
fig.show()
```

#### Nuclei Channels (Blue)

```python
current_blue = None
for x in all_imgs_current:
    if "blue" in x:
        current_blue = x
current_blue = cv2.imread(os.path.join(TRAIN_DIR, current_blue), 0)
fig = px.imshow(current_blue, title='Category: 0, [Nuclei Channels (Blue)]')
fig.show()
```

#### Endoplasmic Reticulum - ER Channels(Yellow)

```python
current_yellow = None
for x in all_imgs_current:
    if "yellow" in x:
        current_yellow = x
current_yellow = cv2.imread(os.path.join(TRAIN_DIR, current_yellow), 0)
fig = px.imshow(current_yellow, title='Category: 0, [ER Channels (Yellow)]')
fig.show()
```

#### Protein of Interest

```python
current_green = None
for x in all_imgs_current:
    if "green" in x:
        current_green = x
current_green = cv2.imread(os.path.join(TRAIN_DIR, current_green), 0)
fig = px.imshow(current_green, title='Category: 0, [Protein of Interest (Green)]')
fig.show()
```
