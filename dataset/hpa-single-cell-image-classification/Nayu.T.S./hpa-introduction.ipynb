Here is a brief description and link for each of the 0~16.

## [Nucleoplasm](https://www.britannica.com/science/nucleus-biology#ref242881)

Main function is to store DNA and facilitate an isolated environment where controlled transcription and gene regulation is enabled. Nucleoplasm is in nucleus.

## [Nuclear membrane](https://en.wikipedia.org/wiki/Nuclear_envelope)

Made up by two lipid bilayer membranes which in eukaryotic cells surrounds the nucleus, which encases the genetic material.

## [Nucleoli](https://en.wikipedia.org/wiki/Nucleolus)

The largest structure in the nucleus of eukaryotic cells. It is best known as the site of ribosome biogenesis.

## [Nucleoli fibrillar center](https://en.wikipedia.org/wiki/Nucleolus#Structure)

Transcription of the rDNA occurs in fibrillar center.

[Picture](https://www.proteinatlas.org/news/2016-07-22/image-of-the-week-nucleoli-fibrillar-center)

## [Nuclear speckles](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC3039535/#:~:text=Nuclear%20speckles%2C%20also%20known%20as,the%20nucleoplasm%20of%20mammalian%20cells.)

Protein found in these speckles serves as a reservoir of factors that participate in transcription and pre-mRNA processing. 


## [Nuclear bodies](https://en.wikipedia.org/wiki/Nuclear_bodies#:~:text=Nuclear%20bodies%20(also%20known%20as,Nuclear%20bodies%20also%20include%20ND10s.))

Membraneless structures found in the cell nuclei of eukaryotic cells.

## [Endoplasmic reticulum](https://en.wikipedia.org/wiki/Endoplasmic_reticulum)

The endoplasmic reticulum serves many general functions, including the folding of protein molecules in sacs called cisternae and the transport of synthesized proteins in vesicles to the Golgi apparatus.

## [Golgi apparatus](https://en.wikipedia.org/wiki/Golgi_apparatus)

The Golgi apparatus is a major collection and dispatch station of protein products received from the endoplasmic reticulum. 

## [Intermediate filaments](https://en.wikipedia.org/wiki/Intermediate_filament)

Play basically a structural role by providing mechanical strength to cells and tissues.

## [Microfilament](https://en.wikipedia.org/wiki/Microfilament)

Protein filaments in the cytoplasm of eukaryotic cells that form part of the cytoskeleton. 

## [Microtubules](https://en.wikipedia.org/wiki/Microtubule)

Polymers of tubulin that form part of the cytoskeleton and provide structure and shape to eukaryotic cells.

## [Mitotic spindle](https://en.wikipedia.org/wiki/Spindle_apparatus)

The cytoskeletal structure of eukaryotic cells that forms during cell division to separate sister chromatids between daughter cells.

## [Centrosome](https://en.wikipedia.org/wiki/Centrosome)

Regulator of cell-cycle progression

## [Plasma membrane](https://www.genome.gov/genetics-glossary/Plasma-Membrane#:~:text=The%20plasma%20membrane%2C%20also%20called,membrane%20on%20its%20outside%20surface.)

The plasma membrane consists of a lipid bilayer that is semipermeable. The plasma membrane regulates the transport of materials entering and exiting the cell by embedded proteins.

## [Mitochondria](https://www.genome.gov/genetics-glossary/Mitochondria)

Generate most of the chemical energy needed to power by the cell's biochemical reactions. 

## [Aggresome](https://en.wikipedia.org/wiki/Aggresome#:~:text=In%20eukaryotic%20cells%2C%20an%20aggresome,proteins%20into%20a%20single%20location.)

An aggregation of misfolded proteins in the cell.

## [Cytosol](https://biologydictionary.net/cytosol/#:~:text=Cytosol%20is%20the%20liquid%20found,of%20life%20to%20take%20place.)

The liquid fills inside of cells.

# Visualize images

```python
def show_train_images(id:str):
    """Show 4 kinds images in train directory.
    """
    
    blue_img = Image.open(DATA + "train/" + f"{id}_blue.png")
    red_img = Image.open(DATA + "train/" + f"{id}_red.png")
    yellow_img = Image.open(DATA + "train/" + f"{id}_yellow.png")
    green_img = Image.open(DATA + "train/" + f"{id}_green.png")
    
    fig, axs = plt.subplots(2, 2,  figsize=(15,15))
    
    axs[0][0].imshow(blue_img)
    axs[0][0].set_title(f"nucleus - {id}")
    axs[1][0].imshow(red_img)
    axs[1][0].set_title(f"microtubules - {id}")
    axs[0][1].imshow(yellow_img)
    axs[0][1].set_title(f"endoplasmic reticulum - {id}")
    axs[1][1].imshow(green_img)
    axs[1][1].set_title(f"the protein of interest - {id}")
```

```python
show_train_images(train.iloc[0]["ID"])
```
