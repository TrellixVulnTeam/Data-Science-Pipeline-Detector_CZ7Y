```python
import plotly.express as px
import matplotlib.pyplot as plt
from plotly.subplots import make_subplots
```

```python
df_train = pd.read_csv("/kaggle/input/hpa-single-cell-image-classification/train.csv")
df_train.shape
```

```python
df_train.head()
```

```python
lbl_cnts = df_train.groupby('Label')['Label'].count().reset_index(name = 'Counts').sort_values(by = 'Counts', ascending = False)
fig = px.bar(lbl_cnts[:20], x='Label', y='Counts', color = 'Label', title = 'Top 20 Label combinations')
fig.show()
```

```python
ls = []
for ele in df_train.Label:
    ls.append(ele.split('|'))
flat_list = [int(item) for sublist in ls for item in sublist]
df = pd.DataFrame(dict((i, flat_list.count(i)) for i in flat_list).items(), columns = ['key','value'])
fig = px.bar(df, x='key', y='value', color = 'key', title = 'Unique Label distribution')
fig.show()
```

**Unique Label distribution is to know the individual counts of each Label. Nucleoplasm(0) and Cytosol(16) are having high number of Cell samples.**
