```python
total = pd.read_csv('../input/hpa-single-cell-image-classification/sample_submission.csv')
total
sample = pd.read_csv('../input/123456/submission.csv')
sample
```

```python
ID = np.array(total['ID'])
ImageWidth = np.array(total['ImageWidth'])
ImageHeight = np.array(total['ImageHeight'])
PredictionString = np.array(total['PredictionString'])
```

```python
idx = np.array(sample['ID'])
predstr = np.array(sample['PredictionString'])
```

```python
for i in range(ID.shape[0]):
    if ID[i] in idx:
        for j in range(idx.shape[0]):
            if ID[i] == idx[j]:
                PredictionString[i] = predstr[j]
                break
```

```python
submission_df = pd.DataFrame({'ID':ID,'ImageWidth':ImageWidth,'ImageHeight':ImageHeight,'PredictionString':PredictionString})
submission_df.to_csv("submission.csv",index=False)
submission_df
```
