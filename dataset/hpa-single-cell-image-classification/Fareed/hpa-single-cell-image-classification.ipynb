# **Lets's Combine channels and visualize images**

```python
import random 
img_list = sorted(os.listdir(os.path.join(ROOT_DIR, "train/")))
fig, ax = plt.subplots(1, 4, figsize = (20, 20))
b = Image.open(os.path.join(ROOT_DIR, "train/"+img_list[0]))
g = Image.open(os.path.join(ROOT_DIR, "train/"+img_list[1]))
r = Image.open(os.path.join(ROOT_DIR, "train/"+img_list[2]))
y = Image.open(os.path.join(ROOT_DIR, "train/"+img_list[3]))

images = [np.dstack((b,g,r)),
       np.dstack((b,r,y)), 
       np.dstack((b,g,y)), 
       np.dstack((g,r,y))]

for i, img in enumerate(images):
    ax[i].imshow(img)
    ax[i].set_title(f"image{i+1}")
    ax[i].axis("off")
plt.show()
```
