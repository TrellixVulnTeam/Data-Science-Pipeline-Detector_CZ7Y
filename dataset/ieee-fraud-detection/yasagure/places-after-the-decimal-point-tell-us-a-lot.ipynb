## how to make the column

```python
import pandas as pd
import matplotlib.pylab as plt
%matplotlib inline
```

```python

folder_path = '../input/'
train_identity = pd.read_csv(f'{folder_path}train_identity.csv', dtype={'TransactionAmt': str})
train_transaction = pd.read_csv(f'{folder_path}train_transaction.csv')


df = pd.merge(train_transaction, train_identity, on='TransactionID', how='left')
```

```python
df["TransactionAmt"]
```

```python
I multipuly 1000 and make int ones.
```

```python
def change(hoge):
    num = 3
    hoge = int(hoge*1000)
    while(hoge % 10 ==0):
        num = num-1
        hoge = hoge /10
    if num<0:
        num = 0
    return num
  
  
df["TransactionAmt_decimal"] = df["TransactionAmt"].map(change)
```

### let's look at it

```python
df["TransactionAmt_decimal"].head()
```

## The relationship between isFraud and decimal places

```python
df.groupby("TransactionAmt_decimal").count()["isFraud"]
```

```python
df.groupby("TransactionAmt_decimal").count()["isFraud"].plot()
plt.show()
```

```python
df.groupby("TransactionAmt_decimal").mean()["isFraud"]
```

```python
df.groupby("TransactionAmt_decimal").mean()["isFraud"].index
```

```python

plt.bar(df.groupby("TransactionAmt_decimal").mean()["isFraud"].index, df.groupby("TransactionAmt_decimal").mean()["isFraud"].values)
```

We can find that the possibility of fraud is high if decimal places are bigger, and most of data have 1 or 2 digits.
