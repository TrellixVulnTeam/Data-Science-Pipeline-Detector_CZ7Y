## Huawei

Huawei has a lot of series of device.
Blade is also one of them.

```python
def change(some):
    if  some != some:
        return np.nan
    elif "Huawei" in some:
        return 1
    elif "Blade" in some:
        return 1
    
    else:
        return 0
```

```python
df["Huawei"] = df["DeviceInfo"].map(change)
```

```python
df["Huawei"] = df['Huawei']*1
```

```python
df.groupby("Huawei").mean().isFraud
```

```python
df.groupby("Huawei").count().isFraud
```

```python
df.groupby("Huawei").sum().isFraud
```
