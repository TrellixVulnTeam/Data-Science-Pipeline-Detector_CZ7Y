## LightGBMでの学習

```python
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.utils.class_weight import compute_sample_weight
```

```python
X_train, X_eval, y_train, y_eval = train_test_split(
    X_train, y_train,
    test_size=0.3,
    shuffle=False
)
```

```python
lgb_params = {
    'objective': 'binary',
    'num_boost_round': 10000,
    'early_stopping_rounds': 10,
    'scale_pos_weight': 1.5,
    'learning_rate':0.1
}
```

```python
lgb_train = lgb.Dataset(
    X_train, y_train,
    weight = compute_sample_weight(class_weight='balanced', y=y_train).astype('float32')
)

lgb_eval = lgb.Dataset(
    X_eval, y_eval,
    reference=lgb_train
)
```

```python
cat_cols = list(X_train.select_dtypes(include="category").columns)

model = lgb.train(
    lgb_params, 
    lgb_train, 
    valid_sets=lgb_eval,
    categorical_feature = cat_cols
)
```

```python
predict_proba = model.predict(X_test, num_iteration=model.best_iteration)
pred = [0 if i < 0.5 else 1 for i in predict_proba]
```

```python
sub["isFraud"] = predict_proba
sub
```

```python
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap, BoundaryNorm
%matplotlib inline
import seaborn as sns
# 部分的なheatmap
def corrmap(features, figsize=(0, 0), bound=95, annot=None):
    if figsize == (0, 0):
        figsize = (len(features)/2, len(features)/3)
    cmap = ListedColormap(['blue', 'aqua', 'whitesmoke', 'yellow', 'red', 'black'])
    bounds = [-100, -bound, -90, 90, bound, 100, 101]
    norm = BoundaryNorm(bounds,cmap.N)
    pd.options.display.float_format = '{:.1f}'.format
    corr_sub = X_train[features].corr() * 100
    display(corr_sub)
    print("")
    print("bound: {}%".format(bound))
    print("")
    f, ax = plt.subplots(figsize=figsize)
    sns.heatmap(corr_sub, cmap=cmap,norm=norm, vmin=min(bounds), vmax=max(bounds), annot=annot)
```
