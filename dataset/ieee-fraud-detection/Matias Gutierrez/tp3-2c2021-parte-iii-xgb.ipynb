# Random/Grid Search

```python
params = {
        'min_child_weight': [1, 5, 10],
        'gamma': [0.5, 1, 1.5, 2, 5],
        'subsample': [0.6, 0.8, 1.0],
        'colsample_bytree': [0.6, 0.8, 1.0],
        'max_depth': [3, 4, 5],
        'scale_pos_weight' : [1,5,18],
        'n_estimators' : [20,50,100]
        }
```

```python
xgb = XGBClassifier(use_label_encoder=False,eval_metric='logloss', seed=6)
```

```python
folds = 10
param_comb = 10

skf = StratifiedKFold(n_splits=folds, shuffle = True, random_state = 1001)

random_search = RandomizedSearchCV(xgb, param_distributions=params, n_iter=param_comb, scoring='roc_auc', n_jobs=4, cv=skf.split(points,labels), verbose=1, random_state=1001 )

random_search.fit(points, labels)
```
