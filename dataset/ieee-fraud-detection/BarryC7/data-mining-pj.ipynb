```python
result3.loc[result3['id_31'].str.contains('samsung', case=False), 'id_31'] = 'samsung'
result3.loc[result3['id_31'].str.contains('chrome', case=False)& result3['id_31'].str.contains('android', case=False), 'id_31'] = 'chrome'
result3.loc[result3['id_31'].str.contains('chrome', case=False), 'id_31'] = 'chrome'
result3.loc[result3['id_31'].str.contains('ie', case=False), 'id_31'] = 'ie'
result3.loc[result3['id_31'].str.contains('edge', case=False), 'id_31'] = 'edge'
result3.loc[result3['id_31'].str.contains('firefox', case=False), 'id_31'] = 'firefox'
#result3.loc[result3['id_31'].str.contains('Android', case=False) | result3['id_31'].str.contains('android', case=False), 'id_31'] = 'android'
result3.loc[result3['id_31'].str.contains('opera', case=False), 'id_31'] = 'opera'
result3.loc[result3['id_31'].str.contains('safari', case=False), 'id_31'] = 'safari'
#for other
result3.loc[~result3['id_31'].isin(['samsung','chrome','ie','edge','firefox','opera','safari','na']), 'id_31'] = 'other'
```

```python
result3.loc[result3['DeviceInfo'].str.contains('iOS', case=False), 'DeviceInfo'] = 'iOS'
result3.loc[result3['DeviceInfo'].str.contains('Trident', case=False), 'DeviceInfo'] = 'Trident'
result3.loc[result3['DeviceInfo'].str.contains('Windows', case=False), 'DeviceInfo'] = 'Windows'
result3.loc[result3['DeviceInfo'].str.contains('MacOS', case=False), 'DeviceInfo'] = 'MacOS'
#for other
result3.loc[~result3['DeviceInfo'].isin(['iOS','Trident','Windows','MacOS','na']), 'DeviceInfo'] = 'other'
```
