## [Google Smartphone Decimeter Challenge 2022](https://www.kaggle.com/competitions/smartphone-decimeter-2022): A look at all of the ground truth maps
### Note: This takes a while to load; it is best to wait until the **Table of Contents** appears on the right hand side before scrolling down.

```python
import numpy as np
import pandas as pd
import folium
from IPython.display import display, Markdown
from pathlib import Path

path        = Path("../input/smartphone-decimeter-2022")
truth_files = (path / 'train').rglob('ground_truth.csv')

for truth_file in truth_files:
    tmp = pd.read_csv(truth_file)
    title_string    = Path(truth_file).parts[4]
    subtitle_string = Path(truth_file).parts[5]
    display(Markdown('## '+ title_string))
    display(Markdown('#### '+ subtitle_string))
    map = folium.Map(location=[tmp.LatitudeDegrees.mean(),
                               tmp.LongitudeDegrees.mean()], 
                               zoom_start=11, 
                               control_scale=True)
    # now add the phone trajectory
    for i in range(len(tmp)):
        folium.Circle(
        location=[tmp.iloc[i]['LatitudeDegrees'],tmp.iloc[i]['LongitudeDegrees']],
            fill_color='blue',
            radius=8,).add_to(map)
    display(map)
```
