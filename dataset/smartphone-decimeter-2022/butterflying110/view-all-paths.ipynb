```python
import pandas as pd
from pathlib import Path

path        = Path("../input/smartphone-decimeter-2022")
truth_files = (path / 'train').rglob('ground_truth.csv')
import plotly.express as px
gts = []
for truth_file in truth_files:
    tmp = pd.read_csv(truth_file)
    
    title_string    = Path(truth_file).parts[3]
    subtitle_string = Path(truth_file).parts[4]
    phonename       = Path(truth_file).parts[5]
    tmp['path']=subtitle_string
    tmp['phone'] =phonename
    gts.append(tmp)
    ground_truth = pd.concat(gts)
fig = px.scatter_mapbox(ground_truth,
                    # Here, plotly gets, (x,y) coordinates
                    lat="LatitudeDegrees",
                    lon="LongitudeDegrees",
                    text='path',

                    #Here, plotly detects color of series
                    color="path",
                    labels="phone",

                    zoom=9,
                    # center={"lat":37.904, "lon":-86.481},
                    height=1200,
                    width=1600)
fig.update_layout(mapbox_style='stamen-terrain')
fig.update_layout(margin={"r": 0, "t": 0, "l": 0, "b": 0})
fig.update_layout(title_text="GPS trafic")
fig.show()
```
