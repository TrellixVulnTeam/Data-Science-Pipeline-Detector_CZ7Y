<h2 style= "background-color:DodgerBlue;color:white; font-weight:bold;font-size:24px;text-align:center;">  Variable Description </h2>

<mark>cp_type : </mark> indicates samples treated with a compound <br>
<mark>cp_time :</mark> treatment duration <br>
<mark>cp_dose :</mark> dose<br>
<mark>g- :</mark> gene expression<br>
<mark>c- :</mark> cell viability<br>

```python
def barPlot(var):
    var_train  = train[var] # get feature
    var_value = var_train.value_counts() # value_counts() count number of categorical variable 
    
    #visualize
    plt.figure(figsize = (8,4))
    plt.bar(var_value.index,var_value)
    plt.xticks(var_value.index, var_value.index.values)
    plt.ylabel('Frequency')
    plt.title(var)
    plt.show()
    print("{}: \n".format(var,var_value))
```

```python
c1 = ["cp_type","cp_time","cp_dose"]
for x in c1:
    barPlot(x)
```
