# Train_target_xxx
Findings:
- nonscored has 403 columns while scored only has 207
- nonscored and scored do NOT contain any same column names (except for sig_id)


I think we ONLY need to analyze scored csv file, and all the MoAs contained in nonscored csv file are not used in the evaluation of this competition.

```python
df_train_scored = pd.read_csv(dir + 'train_targets_nonscored.csv')
df_train_noscored = pd.read_csv(dir + 'train_targets_scored.csv')
```

```python
df_train_scored.head(10)
```

```python
# What MoAs does scored csv contain? And the frequency for a MoA to happen.

tmp = df_train_scored.drop(['sig_id'], axis = 1
                          ).sum(axis=0).sort_values().reset_index()
tmp.columns = ['MoA', '# of 1s']
fig = px.bar(
    tmp.tail(80),
    x = '# of 1s',
    y = 'MoA',
    orientation = 'h',
    height = 1000, 
    width = 800,
)
fig.show()
```

```python
df_train_noscored.head(10)
```

```python
tmp = df_train_noscored.drop(['sig_id'], axis = 1
                          ).sum(axis=0).sort_values().reset_index()
tmp.columns = ['MoA', '# of 1s']
fig = px.bar(
    tmp.tail(80),
    x = '# of 1s',
    y = 'MoA',
    orientation = 'h',
    height = 1000, 
    width = 800,
)
fig.show()
```
