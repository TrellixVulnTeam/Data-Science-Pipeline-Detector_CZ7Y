```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

```python
df = pd.read_csv("../input/lish-moa/train_features.csv")
df.head()
```

```python
cols_feature = [c for c in df.columns if c not in ["sig_id", "cp_type", "cp_time", "cp_dose"]]
```

```python
target = pd.read_csv("../input/lish-moa/train_targets_scored.csv")
target
```

```python
cols_target = [c for c in target.columns if c not in ["sig_id"]]
```

```python
df = pd.merge(df, target, on="sig_id", how="inner")
```

```python
feature = random.sample(cols_feature, 1)[0]
target = random.sample(cols_target, 1)[0]
x = df[["sig_id",feature, target]].sort_values(by=feature).reset_index(drop=True)
x
```

```python
vc = x[target].value_counts().to_dict()
x["target_valuecnt"] = x[target].map(vc)
x
```

```python
    
def show(features, targets):
    col_num = min( len(features), len(targets))

    title_size = 30
    label_size = 30
    fig, axes = plt.subplots(ncols=col_num, figsize=(36, 10), dpi=100, constrained_layout=True)

    for i, (feature, target) in enumerate(zip(features, targets)):
        x = df[["sig_id",feature, target]].sort_values(by=feature).reset_index(drop=True)
        vc = x[target].value_counts().to_dict()
        x["target_valuecnt"] = x[target].map(vc)
        sns.kdeplot(x[feature].values, ax=axes[i])
        
        sns.scatterplot(data= x, x= feature, y=target, color="red" , size="target_valuecnt", ax=axes[i])

        axes[i].set_xlabel('')
        axes[i].set_ylabel('')
        axes[i].tick_params(axis='x', labelsize=label_size)
        axes[i].tick_params(axis='y', labelsize=label_size)
        axes[i].set_title(f"{feature} {target}", size=title_size, pad=title_size)
    plt.show()
    
def random_show(n_rows=10, n_cols=2):
    for i in range(n_rows):
        features = random.sample(cols_feature, n_cols)
        targets = random.sample(cols_target, n_cols)
        show(features, targets)
        
def show_feature(n_rows=10, n_cols=2):
    features = random.sample(cols_feature, 1)*n_cols
    for i in range(n_rows):
        targets = random.sample(cols_target, n_cols)
        show(features, targets)

def show_target(n_rows=10, n_cols=2):
    targets = random.sample(cols_target, 1)*n_cols
    for i in range(n_rows):
        features = random.sample(cols_feature, n_cols)
        show(features, targets)
```
