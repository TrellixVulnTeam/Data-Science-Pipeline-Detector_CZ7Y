```python
from sklearn.model_selection import train_test_split
x = train.drop("sig_id", axis=1)
y = train["sig_id"]
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
#cp_dose_features=  ["cp_dose"]
#cp_type_features = ["cp_type"]
categorical_features = ["cp_dose", "cp_type"]
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot", one_hot, categorical_features)],
                                remainder="passthrough")
transformed_x_train = transformer.fit_transform(x_train)
transformed_x_test = transformer.fit_transform(x_test)

```

```python
 from sklearn import preprocessing 
```

```python
normalizer = preprocessing.Normalizer().fit(transformed_x_train)
```

```python
xtrainnorm = normalizer.transform(transformed_x_train) 
xtestnorm = normalizer.transform(transformed_x_test) 
```

```python
from sklearn.ensemble import RandomForestClassifier
#model= SVC(kernal='rbf')
classifier = RandomForestClassifier(n_estimators = 10, criterion ='entropy', random_state =42)
classifier.fit(transformed_x_train, y_train)
```

```python
features = test.drop("sig_id", axis=1)
```

```python
test["sig_id"] = classifier.predict(test[features])
```
