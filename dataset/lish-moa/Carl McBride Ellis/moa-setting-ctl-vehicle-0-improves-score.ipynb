### Slightly improve your score with these two lines.
Say you read in the `train_features.csv` as `train`, and `test_features.csv` as `test`, *i.e.*

```python
train = pd.read_csv('../input/lish-moa/train_features.csv')
test  = pd.read_csv('../input/lish-moa/test_features.csv')
```

then before doing anything else directly insert these next two lines, which will set all of the `ctl_vehicle` values to 0. (of which there are 1866 in `train_features.csv`, and 358 in `test_features.csv`):

```python
train.at[train['cp_type'].str.contains('ctl_vehicle'),train.filter(regex='-.*').columns] = 0.0
test.at[test['cp_type'].str.contains('ctl_vehicle'),test.filter(regex='-.*').columns] = 0.0
```

Why do this? Because these values represent what are known as *vehicle controls*:
> Vehicle controls refer to negative controls that should not cause significant changes in expression.

From [Connectopedia](https://clue.io/connectopedia/glossary):
> **Controls—Vehicles (VC)**:   Various forms of “inert” experiments performed to determine, by contrast, effects that are specifically due to a perturbation as opposed to responses associated with handling cells or delivering a perturbagen; a type of negative control.

I have found that setting these controls to 0 improved my score by **0.00012**. I know it is not much at the moment, but it may come in handy later...
