This notebook is created to demonstrate the inconsistency in the PCA results.
The inconsistency seems to be more prominent in the higher degree component of the PCA.

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('../input/lish-moa/train_features.csv')
c_cols = df.columns[df.columns.str.startswith('g-')]
df = df[c_cols]
```

```python
pca = PCA(n_components=45, random_state=42)
pca_components = pd.DataFrame(pca.fit_transform(df))
pca_components
```

```python
pca = PCA(n_components=45, random_state=42)
pca.fit(df)
pca_components = pd.DataFrame(pca.transform(df))
pca_components
```
