```python
import sys
import pandas as pd
import numpy as np
import random
```

```python
train_features = pd.read_csv('../input/lish-moa/train_features.csv')
train_targets_scored = pd.read_csv('../input/lish-moa/train_targets_scored.csv')
train_targets_nonscored = pd.read_csv('../input/lish-moa/train_targets_nonscored.csv')

test_features = pd.read_csv('../input/lish-moa/test_features.csv')
sample_submission = pd.read_csv('../input/lish-moa/sample_submission.csv')
```

```python
test_dummy = test_features.copy()
```

```python
test_dummy.head()
```

```python
test_dummy["sig_id"] = test_dummy["sig_id"].apply(lambda s: "dummy" + s[2:])
```

```python
n = len(test_dummy)
np.random.seed(seed=1)
# resample and add noise
for i, col in enumerate(test_dummy.columns):
    if col == "sig_id":
        continue
    elif col in ("cp_type", "cp_time", "cp_dose"):
        test_dummy[col] = test_dummy[col].sample(frac=1, replace=True, random_state=i).to_numpy()
    else:
        test_dummy[col] = test_dummy[col].sample(frac=1, replace=True, random_state=i).to_numpy() + np.random.normal(1, 1, n)
```
