I can see some guys are having problems in submitting Notebook for this competition. I have gone through some of them as well. 
So, here are some tips I wanna share. Hope it helps

**1. Turn the internet option off in your Notebook, before saving. Look right. Just below the output option, you will find settings. The 4th option under settings bar is internet. Turn that off**

**2.There are 2 options for saving. Quick save and "Save & Run All" option. You have to pick the 2nd option, which says "Save and Run All"**

**3. The submission file must be named submission.csv. Nothing else is acceptable. Not even submission1.csv. Nothing else. Only names submission.csv is acceptable.**

**4. TPU accelerators are not acceptable. Make sure you select GPU accelerators**

**5. Highest runtime for GPU accelerator is 120 minutes. Your model shouldn't run more than that, otherwise they will not accept your solution**

Thats all for now. Hope it helps. If you face any other problems, comment below. I will try to look into it.
