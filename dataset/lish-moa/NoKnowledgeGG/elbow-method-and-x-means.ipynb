# Elbow method
* ref : https://en.wikipedia.org/wiki/Elbow_method_(clustering)

* Search the best parameter in k-means' number of clusters
* gene and cell

```python
import warnings
warnings.simplefilter('ignore', FutureWarning)

import numpy as np
import pandas as pd

from tqdm.notebook import tqdm

# + cluster
from sklearn.cluster import KMeans

# plotly
import plotly.express as px
```

```python
r_train = pd.read_csv('../input/lish-moa/train_features.csv')
r_test = pd.read_csv('../input/lish-moa/test_features.csv')
r_train_target = pd.read_csv('../input/lish-moa/train_targets_scored.csv')
r_sub = pd.read_csv('../input/lish-moa/sample_submission.csv')
```

## search kmeans' best number of clusters

## g-

```python
features_g = [name for name in r_train.columns if name.startswith('g-')]

g_SSE = []
for i in tqdm(range(5,250,5)):
    k_train = r_train[features_g].copy()
    k_test = r_test[features_g].copy()
    k_data = pd.concat([k_train, k_test], axis = 0)
    kmeans = KMeans(n_clusters = i, init='k-means++', n_init=10, max_iter=300, tol=1e-04,random_state = 77)
    kmeans.fit(k_data)
    g_SSE.append(kmeans.inertia_)
    
```

```python
df_g = pd.DataFrame({"gene_SSE":g_SSE,
                    'num':list(range(5,250,5))})
df_g
```

```python
fig = px.line(df_g, x = 'num', y = "gene_SSE",
             title = "gene's SSE of some clusters")
fig.show()
```

## c-

```python
features_c = [name for name in r_train.columns if name.startswith('c-')]

c_SSE = []
for i in tqdm(range(2,100,2)):
    k_train = r_train[features_c].copy()
    k_test = r_test[features_c].copy()
    k_data = pd.concat([k_train, k_test], axis = 0)
    kmeans = KMeans(n_clusters = i, init='k-means++', n_init=10, max_iter=300, tol=1e-04,random_state = 77)
    kmeans.fit(k_data)
    c_SSE.append(kmeans.inertia_)
```

```python
df_c = pd.DataFrame({"cell_SSE":c_SSE,
                    'num':list(range(2,100,2))})
df_c
```

```python
fig = px.line(df_c, x = 'num', y = "cell_SSE",
             title = "cell's SSE of some clusters")
fig.show()
```
