# 用語解説
## 作用機序(一般的な意味として)

Wikipediaより、「作用機序」とは以下のように書いてあります。
> 薬理学における作用機序（さようきじょ、英: mechanism of action, MOA）とは、薬剤がその薬理学的効果を発揮するための特異的な生化学的相互作用を意味する[2]。作用機序では大抵、薬剤が結合する酵素あるいは受容体といった特定の分子標的について言及される[3]。受容体部位は、薬物の化学構造とそこで起こる特定の作用に基づき、薬物に対する特定の親和性を持つ

上記の解説は、薬学に詳しい人じゃないと意味を理解できないと思われる<br>
要するに、(薬の)はたらきの仕組みのこと。<br>
例えば、解熱鎮痛剤であるアスピリンを使ったら、痛みを発生させている素を邪魔をする(作用)のことを指す。

## 分子標的薬
分子標的薬は抗癌剤の一種。<br>
「標的」は、コンペの中では「ターゲット」と呼んでいるようだ。<br>
癌の例で言えば、癌の素になる異常なタンパク質が、DNAが傷つくことにより生成される。<br>
異常なタンパク質を標的にして阻害する薬が、分子標的薬のこと。<br>

## 遺伝子発現
遺伝子の情報を元に目的のタンパク質を作るまでの過程を「遺伝子発現」という。<br>
胃癌の例を取ると、遺伝子発現の仕方が１つだけでなく、複数ある。

## 創薬の流れ

中日新聞の「「創薬」って何？」の記事がわかりやすかった。
よって、中日新聞の記事を例に創薬の流れを解説する。
コンペとは直接関係ないかもしれないが、知っていると、比較的戦いやすくなると思います。

創薬には、9-17年の歳月と巨額の研究開発費が必要です。
しかも、大量のコストを掛けても、薬の候補として作り出された物質が新薬になる成功確率は三万分の一とも言われています。

そのような問題を解決するためにAIを導入すると考えられる。
今回のコンペも、創薬に対する問題解決の一環として行われたものと思われます。


- 過程1・標的分子の固定
    - 肺がんの例であれば、肺がんを引き起こしているタンパク質の特定(固定)すること。

- 過程2・候補化合物の探索(スクリーニング)
    - 特定したタンパク質を標的にする化合物(薬の候補)の候補を探す
    - 候補探しは、何億パターンもある鍵穴に対して、合う鍵を探すので膨大な時間とお金がかかるようだ。

- 過程３・候補化合物の最適化
    - 過程２で見つけた化合物の候補から、本当に使えそうな化合物(薬)を最適化すること。

- 過程4・臨床試験
    - 動物実験→人で実験すること

- 過程5・申請承認
    - 承認請求

# コンペで使用するデータを覗いてみる

- train_features.csv - Features for the training set. Features g- signify gene expression data, and c- signify cell viability data. cp_type indicates samples treated with a compound (cp_vehicle) or with a control perturbation (ctrl_vehicle); control perturbations have no MoAs; cp_time and cp_dose indicate treatment duration (24, 48, 72 hours) and dose (high or low).
    - 「g-」は遺伝子発現データ。
    - 「ｃ−」は、細胞生存率データ。カラムが丁度100個ある。
    - cp_typeは、化合物（cp_vehicle）または制御摂動（ctrl_vehicle）で処理されたサンプル。ctrl_vehicleには、作用機序はない。
    - cp_timeは、翻訳すると治療期間になる。この治療期間とは、１回の投薬に対しての期間？
    - doseは、翻訳すると投与量。

- train_drug.csv - This file contains an anonymous drug_id for the training set only.
    - train_features.sig_idとtrain_drug.sig_idで紐づく。train_drugの下にtrain_featuresがあるイメージ

- train_targets_scored.csv - The binary MoA targets that are scored.
   - ターゲットになる薬のデータという解釈


```python
import pandas as pd

# ファイルへのアクセス設定
ROOT = '../input/lish-moa/'

# ファイルのインポート
train_features = pd.read_csv(ROOT + 'train_features.csv')
train_targets_scored = pd.read_csv(ROOT + 'train_targets_scored.csv')

pd.set_option('display.max_columns', 1000)
```

```python
train_targets_scored
```

```python
#ターゲットとなる薬を全て出力してみる
#例えば、出力結果にある「akt_inhibitor」は、「Akt阻害薬」のことで、乳がん治療の新薬として注目されているようだ。

target_list = list(train_targets_scored.columns)
target_list.remove('sig_id')
target_list
```

```python
#akt_inhibitorに絞って訓練データ、ターゲット情報を見る
train_targets_scored_akt_inhibitor =  train_targets_scored[train_targets_scored["akt_inhibitor"] == 1]
train_targets_scored_akt_inhibitor
```

```python
sig_id = train_targets_scored_akt_inhibitor["sig_id"].values.tolist()
train_features[train_features['sig_id'].isin(sig_id)]
```
