```python
import pandas as pd
import numpy as np
```

```python
feats = pd.read_csv('../input/lish-moa/train_features.csv', index_col='sig_id')
drugs = pd.read_csv('../input/lish-moa/train_drug.csv', index_col='sig_id')
y = pd.read_csv('../input/lish-moa/train_targets_scored.csv', index_col='sig_id')
```

```python
drugs.value_counts()
```

Does each drug have the same MoA, no matter the condition?

```python
moa = pd.Series(data=list(y.values), index=y.index, name='moa')
moa_with_drug = moa.to_frame().merge(drugs, how='inner', right_index=True, left_index=True)
```

```python
def match(df):
    if len(df) == 1:
        return True
    x = np.stack(df['moa'].values)
    return np.array_equal(np.max(x, axis=1), np.min(x, axis=1))
    
moa_with_drug.groupby('drug_id').apply(match).value_counts()
```

Strange, I guess not! The mechanism of action may differ based on dosage and time.
