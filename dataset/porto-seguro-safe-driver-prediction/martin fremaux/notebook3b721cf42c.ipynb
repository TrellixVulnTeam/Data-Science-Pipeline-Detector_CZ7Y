## La m√©thode random forest classifier semble efficace et rapide, on la choisit donc pour la suite

```python
X_train = train.drop(['target'], axis=1)
y_train = train.target
```

```python
smk = SMOTETomek()
X_res, y_res = smk.fit_resample(X_train, y_train)
```

```python
X_train_res, X_test_res, y_train_res, y_test_res = train_test_split(X_res, y_res, test_size=0.2, random_state=42)
```

```python
X_res.head
```

```python
import sklearn

rf = sklearn.ensemble.RandomForestClassifier()
```

```python
rf.fit(X_train_res, y_train_res)
```

```python
y_rf = rf.predict(X_test_res)
```

```python
from sklearn.metrics import * 

confusion_matrix(y_test_res,y_rf)
```

```python
y_test_tolist = y_test_res.tolist()
print("Train 0 : " + str(y_test_tolist.count(0)))
print("Train 1 : " + str(y_test_tolist.count(1)))

y_rf_tolist = y_rf.tolist()
print("Test 0 : " + str(y_rf_tolist.count(0)))
print("Test 1 : " + str(y_rf_tolist.count(1)))
```

```python
predict_submission = rf.predict(test)
```

```python
submission = pd.DataFrame({'id': id,'target': predict_submission})
```
