```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.ensemble import RandomForestClassifier

# Read in our input data
df_train = pd.read_csv('/kaggle/input/porto-seguro-safe-driver-prediction/train.csv')
df_test = pd.read_csv('/kaggle/input/porto-seguro-safe-driver-prediction/test.csv')

y_train = df_train['target'].values
id_train = df_train['id'].values
id_test = df_test['id'].values

x_train = df_train.drop(['target', 'id'], axis=1)
x_test = df_test.drop(['id'], axis=1)
```

```python
x_train, x_valid, y_train, y_valid = train_test_split(x_train, y_train, test_size=0.2, random_state=4242)
print('Train samples: {} Validation samples: {}'.format(len(x_train), len(x_valid)))
```

```python
model = RandomForestClassifier(class_weight="balanced", min_samples_split=10, n_jobs=-1)
```

```python
model.fit(x_train, y_train)
```

```python
yhat = model.predict(x_valid)
```

```python
cm = confusion_matrix(y_valid, yhat)
ConfusionMatrixDisplay(cm).plot();
```

```python
yhat_test = model.predict(x_test)
```

```python
sub = pd.DataFrame()
sub['id'] = id_test
sub['target'] = yhat_test
sub.to_csv('predictions.csv', index=False)
```
