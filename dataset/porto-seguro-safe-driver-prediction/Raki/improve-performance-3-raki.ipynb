# 성능개선 3 : LightGBM 과 XGBoost 앙상블

> 앙상블에는 다양하 모델을 사용하는 게 바람직함. 각 모델의 예측값들이 거의 같다면 앙상블을 해도 큰 효과가 없음. 반면 타깃값 0을 잘 예측하는 모델과 타깃값 1을 잘 예측하는 모델을 앙상블하면 시너지 효과를 발휘해 성능이 더 높아질 가능성이 큼. 다양한 모델로 앙상블해야 성능이 좋아질 가능성이 더 높다는 점 꼭 기억!

```python
import pandas as pd
```

```python
# 데이터 경로
data_path = '/kaggle/input/'

lgb_submission = pd.read_csv(data_path + 'improve-performance-1-raki/submission.csv', index_col='id')
xgb_submission = pd.read_csv(data_path + 'improve-performance-2-raki/submission.csv', index_col='id')
submission = pd.read_csv(data_path + 'porto-seguro-safe-driver-prediction/sample_submission.csv', index_col='id')
```
