```python
test=pd.read_csv("../input/porto-seguro-safe-driver-prediction/test.csv")
train=pd.read_csv("../input/porto-seguro-safe-driver-prediction/train.csv")
```

```python
train.head()
```

```python
y_train=train["target"]
X_train=train.drop(["target"],axis=1)
```

```python
from supervised.preprocessing.eda import EDA 
EDA.extensive_eda(X_train,y_train,save_path=".")
```

```python
y_train.value_counts()
```

```python
from imblearn.under_sampling import RandomUnderSampler 
rus = RandomUnderSampler()
X_train, y_train = rus.fit_resample(X_train, y_train)
```

```python
y_train.value_counts()
```

```python
from supervised.automl import AutoML 
automl = AutoML(mode="Explain",eval_metric="accuracy")
```

```python
automl.fit(X_train,y_train)
```

```python
automl.report()
```

```python
predictions=automl.predict(test)
```

```python
sample_submission=pd.read_csv("../input/porto-seguro-safe-driver-prediction/sample_submission.csv")
```

```python
sample_submission["target"]=predictions
```

```python
sample_submission.to_csv("submisson.csv",index=False)
```
