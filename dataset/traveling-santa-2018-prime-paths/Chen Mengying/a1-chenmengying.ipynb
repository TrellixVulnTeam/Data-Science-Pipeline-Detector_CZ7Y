<h1>Final version</h1>
<p>
    For the final one, a fixed point will be set, it calculate the distence between the fixed city point and all the other cities point, after getting the distance between, the distance will be compared, then we come out with the closest distance. Then the current fixed city point will be removed, and will be changed to the point which is has the closest distance. Then this  calculation will loop until all the point was calculated. Add the distances up into the total distance. Besides, the cities which be removed will be append, and for showing the order.
</p>

<p>
    By using this way, we can make sure that the every distance is the shortest.
</p>

<p>
    For the data structure, linked list and queue was used.<br>
    The linked list was used to stored the cities data from csv file and be used to the calculation, and queue list was used to store the order of the cities it went.
</p>

<p>
    Compared with the normal list, it is quicker to insert annd remove the items in middle of a big list. It both takes O(1) time to add and remove items. However, for the regular list, it will takes same time to append items, but longer time to remove. which use O(n) time.
</p>

<p>
    For using queue instead of using stack is beacuse queue is FIFO (first in first out), stack is LIFO(last in first out). The way be used should be the first city be removed(be calculated) go in to list first, and when plot in graph, the first point show first (pop out), then is other.
</p>

<h3>Queue Class</h3>

```python
import collections

class Queue:
    def __init__(self, *args, **kwargs):
        self._data = [] #empty list


    #PUSH operation
    def push(self, item):
        self._data.append(item)


    #POP operation
    def pop(self):
        if not self.is_empty():
            item = self._data.pop(0)
            return item
        else:
            print("Error: Queue empty")


    def is_empty(self):
        if len(self._data) == 0:
            return True
        else:
            return False

    def display(self):
        print("Order of cities: ", self._data)


    def top(self):
        if not self.is_empty():
            return self._data[-1]
        else:
            print("Error: Queue empty")
```
