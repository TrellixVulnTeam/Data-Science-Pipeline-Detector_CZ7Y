**Prime Cities Calculation**

Determines whether integers are prime by producing a list of all numbers, this is done by marking numbers that aren't prime and the multiples of each starting at 2. This determines which cities are prime and which aren't

```python
def sieve_of_eratosthenes(n):
    primes = [True for i in range (n + 1)]
    primes[0] = False
    primes[1] = False
    for i in range(2, int(np.sqrt(n)) + 1):
        if primes[i]:
            k = 2
            while i * k <= n:
                primes[i * k] = False
                k += 1
    return(primes)
prime_cities = sieve_of_eratosthenes(max(cities.CityId))
```

**Merge Sort Algorithm**

```python
def merge(S1, S2, S):
    i = j = k = 0
    while i < len(S1) and j < len(S2):
        if Y[S1[i]] < Y[S2[j]]:
            S[k] = S1[i]
            i += 1
        else:
            S[k] = S2[j]
            j += 1
        k += 1
```

```python
def merge_sort(S):
    n = len(S)
    if len(S) > 1:
        mid = len(S) // 2
        S1 = S[0:mid]
        S2 = S[mid:n]
        merge_sort(S1)
        merge_sort(S2)
        merge(S1, S2, S)

mergesort_path = []
for x in range(1, max(cities.CityId) + 1):
    mergesort_path.append(x)
merge_sort(mergesort_path)

sorted_mergsort_path = [0]
for i in range(len(mergesort_path)):
    sorted_mergsort_path.append(mergesort_path[i])
sorted_mergsort_path.append(0)

print('Total distance traveled using merge sort: {:,}'.format(total_distance(cities, sorted_mergsort_path)))
```

The worst run time of this algorithm is constant, O(n). The biggest operation present is the iteration of the for loop

```python
def quick_sort(arr, low, high):
    if low < high:
        
        # pi partitions index, arr[p] is now at right place
        pi = partition(arr, low, high)
        
        #Separately sort elements before partitioning and after partitioing
        quick_sort(arr, low, pi - 1)
        quick_sort(arr, pi + 1, high)
```
