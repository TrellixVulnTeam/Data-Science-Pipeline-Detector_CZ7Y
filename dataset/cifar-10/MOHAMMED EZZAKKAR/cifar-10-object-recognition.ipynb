# <h3>Téléchargement et prétraitement des donnees :<h3>

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from keras.datasets import cifar10
from keras.utils.np_utils import to_categorical
```

```python
data=cifar10.load_data()
```

```python
(x_train, y_train), (x_test, y_test) = data
```

```python
print(x_train.shape, 'train shape')
print(x_test.shape, 'test shape')
print(y_train.shape, 'train samples')
print(y_test.shape, 'test samples')
```

```python
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
plt.figure(figsize=(16,16))
for i in range(5):
    plt.subplot(6,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(x_train[i])
    plt.xlabel(class_names[y_train[i][0]])
plt.show()
```

```python
#Normalisation
x_train = x_train.astype('float32')
x_train/=255
x_test = x_test.astype('float32')
x_test/=255

#convertir en un vecteur de 10 qui a des valeurs binaires et des colonnes égales au nombre de catégories dans les données.
#exemple : 2 --> 0100000000
num_classes = 10
y_train_ohe = to_categorical(y_train, num_classes) 
y_test_ohe = to_categorical(y_test, num_classes)
```

# <h2>La création de modéle d'apprentissage :</h2>
<h3>l'archetecture CNN :</h3>
<h4>Au cours des dernières années, Deep Learning a prouvé qu'il s'agissait d'un outil très puissant en raison de sa capacité à gérer d'énormes quantités de données. L'utilisation de couches cachées dépasse les techniques traditionnelles, en particulier pour la reconnaissance de formes. L'un des réseaux de neurones profonds les plus populaires est le réseau de neurones convolutifs (CNN).<h4>
<img src='https://i.morioh.com/200516/91e28a8b.jpg'/>

```python
from keras.models import Sequential
from keras.layers import Dense, Conv2D, MaxPooling2D, Flatten, Dropout
from keras.activations import relu, softmax
from keras.layers.normalization import batch_normalization
from keras import layers, activations
```

<h4>Notre modele CNN se compose de:<br><br><br>
I-Réseau des neuronnes convolutif de 4 layers chacune contient :
<br><br>1- Une couche de convolution 2D avec la methode Conv2D ou il applique des filtres aléatoirement en fixant leurs nombre et dimention(layer1 :64 filtres de (2,2))
<br>2- Une couche de normalisation par batch avec la methode BatchNormalization() qui normalise chaque batch de données indépendamment. Pour accélérer le training du réseau neuronal convolutif et réduire la sensibilité à l'initialisation du réseau.
<br>3- Puit l'application de fonction d'activation relu qui elimine les valeurs negatives et assuré la non linearité de notre réseau.
<br>4- Et on termine par la compression des images on appliquant le pooling pour réduire leur taille iniciale en gardent les valeurs les plus important via la methode MaxPooling2D.

<br><br>II-Un Perceptron Multicouch :<br><br>
   1- Appliquer la fonction Flatten qui transférer les entrées multidimensionnels en une seule dimension.<br>
   2- Puis trois couche lineaire chacune suivi par une fonction d'activation relu<br>
   3- On appliquant la fonction dropout qui est une technique de régularisation ou on supprime un pourcentage des neurones aleatoirement de maniére temporaireet pendant le training les autres neurones faire leurs prédiction pour obtien un réseau moins sensible au poid specifique des neurones.</h4>

```python
model_=Sequential()
model_.add(Conv2D(64, kernel_size=(2,2), input_shape=(32, 32, 3), padding='same'))
model_.add(batch_normalization.BatchNormalization())
model_.add(layers.Activation(activations.relu))
model_.add(MaxPooling2D(pool_size=(2,2)))

model_.add(Conv2D(64, kernel_size=(2,2), padding='same'))
model_.add(batch_normalization.BatchNormalization())
model_.add(layers.Activation(activations.relu))
model_.add(MaxPooling2D(pool_size=(2,2)))

model_.add(Conv2D(128, kernel_size=(2,2), padding='same'))
model_.add(batch_normalization.BatchNormalization())
model_.add(layers.Activation(activations.relu))
model_.add(MaxPooling2D(pool_size=(2,2)))

model_.add(Conv2D(128, kernel_size=(2,2), padding='same'))
model_.add(batch_normalization.BatchNormalization())
model_.add(layers.Activation(activations.relu))
model_.add(MaxPooling2D(pool_size=(2,2)))




model_.add(Flatten())
model_.add(Dense(512))
model_.add(Dropout(0.25))
model_.add(layers.Activation(activations.relu))
model_.add(Dense(10, activation='softmax'))


model_.summary()
```

# La Compilation et L'Apprentissage du Modèle :

```python
from keras.callbacks import EarlyStopping, ModelCheckpoint
```

```python
es = EarlyStopping(monitor='val_accuracy', mode='max', patience=20 )# cad si la valeur minimale d'erreur ne diminue pas même après 20 epoche le programe  s'arrête
mc=ModelCheckpoint('./weights1.h5', save_best_only=True)#on enregistre les régles qui nous permet d'obtenir la meilleure prediction  avec le  minimum des erreurs dans weights1.h5
model_.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])#Nous précisons la configuration de training #optimizer: optimise les poids d'entrée en comparant la prédiction et la fonction de perte #loss : c'es fct d'erreur #metrics: évaluer les performances de modèle during training and testing
h1=model_.fit( x_train, y_train_ohe, batch_size=100, epochs=50, validation_data= (x_test,y_test_ohe ), callbacks=[es, mc])#Trains the model 
```

```python
model_.load_weights("weights1.h5")
model_.evaluate(x_test,y_test_ohe)
```

```python
plt.plot(h1.history['accuracy'])
plt.plot(h1.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
# summarize history for loss
plt.plot(h1.history['loss'])
plt.plot(h1.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

# **Transfert d'apprentissage : The Fine Tuning VGG16(imagenet)**

<h4>Transfert d'apprentissage est une méthode d'apprentissage automatique où l'application des connaissances obtenues à partir d'un modèle utilisé dans une task peut être réutilisée comme point de départ pour une autre task.
dans ce qui suit on va appliqué le transfert d'apprentissage de modéle VGG16 entrainé sur ImageNet à notre dataset</h4>
<h4>Puisque notre dataset est un peut different de ImageNet dataset et petite le Fine Tuning nous permettra généralement d'obtenir un meilleur résultat. C'est le meme que transfert d'apprentissage sauf qu'on peut retrain the layers of the feature extraction.</h4>

```python
from keras.applications.vgg16 import VGG16
from keras.utils.np_utils import to_categorical
from keras.applications.vgg16 import preprocess_input
```

```python
(x_train, _), (x_test, _) = data


x_train=preprocess_input(x_train)#convertira les images d'entrée de RVB en BGR, puis centrera à zéro chaque canal de couleur par rapport au ImageNet dataset.
x_train/=255

x_test=preprocess_input(x_test)
x_test/=255


#x_train[0]
print(x_train[0])
```

```python
import tensorflow as tf
vggmdl = tf.keras.applications.VGG16(include_top=False, weights='imagenet', input_shape=(32,32,3))

model = Sequential()
for layer in vggmdl.layers:
    model.add(layer)


for layer in model.layers[-7:]:
    layer.trainable = False

    
for layer in model.layers:
    print(layer.trainable)
```

```python
model.add(Flatten())

model.add(Dense(512))
model.add(Dropout(0.25))
model.add(layers.Activation(activations.relu))
#model.add(Dense(512, activation='relu'))
model.add(Dense(10, activation='softmax'))


model.summary()
```

```python
es = EarlyStopping(monitor='val_loss', mode='min', patience=20 )
mc=ModelCheckpoint('./weights2.h5', save_best_only=True)
model.compile(optimizer='adam', loss='categorical_crossentropy',metrics='accuracy')
h2=model.fit( x_train, y_train_ohe, batch_size=100, epochs=50, validation_data= (x_test,y_test_ohe ), callbacks=[es, mc])
```

```python
model.load_weights("weights2.h5")
model.evaluate(x_test,y_test_ohe)
```

```python
plt.plot(h2.history['accuracy'])
plt.plot(h2.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
# summarize history for loss
plt.plot(h2.history['loss'])
plt.plot(h2.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
