```python
# importing data sets and other libraries
from keras.datasets import cifar10
from keras import models
from keras import layers
from keras import Sequential
```

```python
# global parameters'
epochs=10
```

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

```python
# merging the training and test sets so that we can apply cross validation
X_combined = np.r_[x_train, x_test]
y_combined = np.r_[y_train, y_test]
```

```python
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import Conv2D, MaxPooling2D
```

```python
# norimalising the data 
X_combined = X_combined.astype('float32')
y_combined = y_combined.astype('float32')
X_combined /= 255
y_combined /= 255
```

```python
def build_model():
    model=Sequential()
    model.add(Conv2D(32, (3, 3), padding='same',input_shape=x_train.shape[1:]))
    model.add(Activation('relu'))
    model.add(Conv2D(32, (3, 3)))
    model.add(Activation('relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(layers.Flatten())
    model.add(layers.Dense(512, activation='relu'))
    model.add(layers.Dense(1, activation='sigmoid'))
    model.add(Dropout(0.25))
    model.compile(optimizer='rmsprop', loss='mse', metrics=['accuracy'])
    return model
    
    
    
```

```python
# test 
# manual k fold using fc 
k=10
validation_samples=int(len(X_combined)/k)

validation_scores=[]
number_of_epochs=20


```

```python
validation_samples
```

```python
for i in range(k):
    print('processing fold #', i)
    x_test= X_combined[i * validation_samples: (i + 1) * validation_samples]
    y_test = y_combined[i * validation_samples: (i + 1) * validation_samples]
    x_train = np.concatenate([X_combined[:i * validation_samples],X_combined[(i + 1) * validation_samples:]],axis=0)
    y_train = np.concatenate([y_combined[:i * validation_samples],y_combined[(i + 1) * validation_samples:]],axis=0)
    model = build_model()
    model.fit(x_train, y_train,epochs=2, batch_size=1, verbose=2)
    val_mse, val_mae = model.evaluate(x_test, y_test, verbose=2)
    validation_scores.append(val_mae)
```
