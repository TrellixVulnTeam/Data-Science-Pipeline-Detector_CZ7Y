```python
#Manipulateur des tableaux
import numpy as np

# Manipulation des images| OpenCV
import cv2

# Afficher les images et evaluer le model résultat
import matplotlib.pyplot as plt

# Model VGG16  
from keras.applications.vgg16 import VGG16

# Preparation du model VGG16
from keras.layers import Dense,Flatten,Input
from keras.models import Sequential

# One hot label encoding
from keras.utils import np_utils

# Base de donnee CIFAR10
from keras.datasets import cifar10
```

```python
#Chargement de donnee

(x_train,y_train),(x_test,y_test) = cifar10.load_data()
print("Shape of x_train is ",x_train.shape)
print("Shape of y_train is ",y_train.shape)
print("Shape of x_test  is ",x_test.shape)
print("Shape of y_test  is",y_test.shape)
```

```python
# one hot encoding
y_train =  np_utils.to_categorical(y_train,num_classes=10)
y_test =  np_utils.to_categorical(y_test,num_classes=10)

print("New shape of y_train is ",y_train.shape)
print("New shape of y_test  is ",y_test.shape)
```

```python
# Include top = ajouter des couches entièrement connectés au couche
# Weights = utilise  weights deja entainer(trained in imagenet)
vgg = VGG16(include_top=False,weights="imagenet",input_shape=(32,32,3))

vgg.summary()
```

```python
#Creation d'un model 
model = Sequential()

# Ajouter les couches au modèle blank
for layer in vgg.layers:
    model.add(layer)
    
#N'entraînez plus les couches, car elles sont déjà entraînées
for layer in model.layers:
    layer.trainable = False
    
#Ajout de couches entièrement connectées

model.add(Flatten())
model.add(Dense(units=4098, activation='relu', name='fc1')) 
model.add(Dense(units=4098, activation='relu', name='fc2'))
model.add(Dense(units=10, activation='softmax', name='predictions'))

# verifier le model
model.summary()
```

```python
# Compiler le model
model.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])

# Lancer l'entraînement.
#Nous utilisons %15 du data d'entrainement comme ensemble de validation.
hist = model.fit(x_train,y_train,epochs=20,batch_size=68, validation_split =0.15)
```

```python
import cv2
IMG_SIZE=32
path1='../input/test-data/cat.jpeg'
#path1='../input/tst123/mix.jpg'
img_data1 = cv2.imread(path1)
from matplotlib import pyplot as plt
plt.imshow(img_data1)

img_data1 = cv2.resize(img_data1, (IMG_SIZE, IMG_SIZE))
data1 = img_data1.reshape(-1, IMG_SIZE, IMG_SIZE, 3)


model_out=model.predict(data1)
print(model_out)


if np.argmax(model_out) == 1:
    str_label = 'Automobile'
    print(str_label)
if np.argmax(model_out) == 2:
    str_label = 'Bird'
    print(str_label)
if np.argmax(model_out) == 3:
    str_label = 'Cat'
    print(str_label)
if np.argmax(model_out) == 4:
    str_label = 'Deer'
    print(str_label)
if np.argmax(model_out) == 0:
    str_label = 'Airplane'
    print(str_label)
if np.argmax(model_out) == 5:
    str_label = 'Dog'
    print(str_label)
if np.argmax(model_out) == 6:
    str_label = 'Frog'
    print(str_label)
if np.argmax(model_out) == 7:
    str_label = 'Horse'
    print(str_label)
if np.argmax(model_out) == 8:
    str_label = 'Ship'
    print(str_label)
if np.argmax(model_out) == 9:
    str_label = 'Truck'
    print(str_label)
```

```python
test_scores = model.evaluate(x_test, y_test, verbose=2)
print("Test loss:", test_scores[0])
print("Test accuracy:", test_scores[1])
```

```python
predict_y = model.predict(x_test)
predict_y = np.round(predict_y)
```

```python
labels = [
    "airplane",
    "automobile",
    "bird",
    "cat",
    "deer",
    "dog",
    "frog",
    "horse",
    "ship",
    "truck"
]
```
