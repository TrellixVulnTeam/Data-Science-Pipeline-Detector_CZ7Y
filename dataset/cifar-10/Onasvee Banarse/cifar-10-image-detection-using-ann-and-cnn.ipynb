```python
import os 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import zipfile
import tensorflow as tf
from tensorflow.keras.optimizers import RMSprop
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import datasets,layers,models,Model
```

```python
# loading data set

(X_train,Y_train),(X_test,Y_test)=datasets.cifar10.load_data()
```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
# reshape  2 d to 1 d 
Y_train = Y_train.reshape(-1,)
Y_test = Y_test.reshape(-1,)
```

```python
classes = ["airplane","automobile","bird","cat","deer","dog","frog","horse","ship","truck"]
```

```python
def plot_sample(X, Y, index):
    plt.figure(figsize = (15,2))
    plt.imshow(X[index])
    plt.xlabel(classes[Y[index]])
```

```python
plot_sample(X_train, Y_train, 6)
```

```python
for i in range(10):
  plot_sample(X_train, Y_train, i)
```

```python
# normalization
X_train=X_train/255
X_test=X_test/255
```

Simple ANN

categorical_crossentropy= when y is one hot encoded 0000100
sparse_categorical_crossentropy = when y is a define value y=9

```python
model_ann = models.Sequential([
        layers.Flatten(input_shape=(32,32,3)),
        layers.Dense(3000, activation='relu'),
        layers.Dense(2000, activation='relu'),
        layers.Dense(1000, activation='relu'),
        layers.Dense(10, activation='softmax')    
    ])

model_ann.compile(optimizer='SGD',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

MODEL_1=model_ann.fit(
    X_train, Y_train, epochs=10
)
```

ANN=**accuracy: 0.5764**

**CNN**

```python
model_cnn = models.Sequential([
        layers.Conv2D(filters=32,kernel_size=(3,3),activation='relu',input_shape=(32,32,3)),
        layers.MaxPooling2D((2,2)),
        
        layers.Conv2D(filters=64,kernel_size=(3,3),activation='relu'),
        layers.MaxPooling2D((2,2)),
        layers.Dropout(0.5),
    
        layers.Conv2D(filters=128,kernel_size=(3,3),activation='relu'),
        layers.MaxPooling2D((2,2)),
        layers.Dropout(0.5),
        
        layers.Flatten(),
        layers.Dense(512, activation='relu'),
        layers.Dense(10, activation='softmax')    
    ])
```

```python
model_cnn.compile(
                loss = 'sparse_categorical_crossentropy',
               optimizer = RMSprop(learning_rate=1e-4),
               metrics = ['accuracy'])


MODEL_2=model_cnn.fit(X_train,Y_train,
                        epochs=20,
                      verbose=1)
```

```python
model_cnn.evaluate(X_test,Y_test)
```

CNN_RMSprop=
* **accuracy on train: 0.6262**
* **accuracy on test: 0.6644**

```python
model_cnn_2 = models.Sequential([
        layers.Conv2D(filters=32,kernel_size=(3,3),activation='relu',input_shape=(32,32,3)),
        layers.MaxPooling2D((2,2)),
        
        layers.Conv2D(filters=64,kernel_size=(3,3),activation='relu'),
        layers.MaxPooling2D((2,2)),
        layers.Dropout(0.5),
    
        layers.Conv2D(filters=128,kernel_size=(3,3),activation='relu'),
        layers.MaxPooling2D((2,2)),
        layers.Dropout(0.5),
        
        layers.Flatten(),
        layers.Dense(512, activation='relu'),
        layers.Dense(10, activation='softmax')    
    ])

model_cnn_2.compile(
                loss = 'sparse_categorical_crossentropy',
               optimizer = "adam",
               metrics = ['accuracy'])


MODEL_3=model_cnn_2.fit(X_train,Y_train,
                        epochs=20,
                      verbose=1)
```

```python
model_cnn_2.evaluate(X_test,Y_test)
```

CNN_adam=
* **accuracy on train: 0.6880**
* **accuracy on test: 0.7342**
