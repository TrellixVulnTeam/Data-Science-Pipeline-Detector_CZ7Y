```python
import os
os.listdir('../input')
```

```python
import numpy as np, pandas as pd

f_attn_bi = '../input/module-12-attn-bi/10fold_attn_post_am (1).csv'
f_dpcnn = '../input/module-10-dpcnn/10fold_dpcnn_test (1).csv'
f_nbsvm = '../input/module-1-nbsvm/Module_1_submission.csv'
f_bilstm_f_pl = '../input/module-2-bilstm-f-pl/Module_2_submission.csv'
f_capsule_gru = '../input/module-7-capsule-gru/Module_7_submission.csv'
f_mish = '../input/module-6-mish/Mish_submission.csv'
f_bilstm_g_tta = '../input/module-3-bilstm-g-tta/Module_3_submission.csv'
f_capsule = '../input/module-11-capsule/10fold_capsule_am.csv'
f_dmcnn = '../input/module-9-dmcnn/10fold_dmcnn_am.csv'
f_bilstm_f = '../input/module-13-bilstm-f/Module_13_submission.csv'
f_dehyp = '../input/module-4-dehyph/Module_4_submission.csv'
f_bilstm_duembd_pl = '../input/module-5-bi-lstm-duembd-pl/Module_5_submission.csv'
f_bilstm_pp = '../input/module-8-bi-lstm-pp/10fold_lstmpp_am (1).csv'
```

```python
p_attn_bi = pd.read_csv(f_attn_bi)
p_dpcnn = pd.read_csv(f_dpcnn)
p_nbsvm = pd.read_csv(f_nbsvm)
p_bilstm_f_pl = pd.read_csv(f_bilstm_f_pl)
p_capsule_gru = pd.read_csv(f_capsule_gru)
p_mish = pd.read_csv(f_mish)
p_bilstm_g_tta = pd.read_csv(f_bilstm_g_tta)
p_capsule = pd.read_csv(f_capsule)
p_bilstm_f = pd.read_csv(f_bilstm_f)
p_dmcnn = pd.read_csv(f_dmcnn)
p_dehyp = pd.read_csv(f_dehyp)
p_bilstm_duembd_pl = pd.read_csv(f_bilstm_duembd_pl)
p_bilstm_pp = pd.read_csv(f_bilstm_pp)
```

Now we can take the average of the label columns.

```python
label_cols = ['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']
p_res = p_attn_bi.copy()
p_res[label_cols] = (p_attn_bi[label_cols] + p_dpcnn[label_cols] + p_nbsvm[label_cols] + p_bilstm_f_pl[label_cols] + p_capsule_gru[label_cols] + p_mish[label_cols] + p_bilstm_g_tta[label_cols] + (p_capsule[label_cols] * 3) + p_bilstm_f[label_cols] + p_dmcnn[label_cols] + p_dehyp[label_cols] + p_bilstm_duembd_pl[label_cols] + p_bilstm_pp[label_cols]) / 15
```

And finally, create our CSV.

```python
p_res.to_csv('submission.csv', index=False)
```
