```python
cols = ['toxic','severe_toxic','obscene','threat','insult','identity_hate']
for c in cols:
    print(train[c].sum())
```

```python
cnt=0
for ind in train.index:
    add=0
    for c in cols:
        add = add+train[c][ind]
    if add>1:
        cnt = cnt+1

print(cnt)
```

```python
labels = []
for ind in train.index:
    add=0
    for c in cols:
        add = add+train[c][ind]
    labels.append(add)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(16,9))
sns.countplot(labels)
plt.xlabel('target per comment')
plt.ylabel('Count of target')
plt.show()
```
