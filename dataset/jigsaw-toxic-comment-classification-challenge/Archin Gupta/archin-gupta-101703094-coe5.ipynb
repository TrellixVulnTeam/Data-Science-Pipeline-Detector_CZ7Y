```python
import numpy as np, pandas as pd
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras.layers import Dense, Input, LSTM, Embedding, Dropout,GlobalMaxPool1D
from keras.models import Sequential
```

```python
train_data = pd.read_csv('../input/jigsaw-toxic-comment-classification-challenge/train.csv.zip')
test_data = pd.read_csv('../input/jigsaw-toxic-comment-classification-challenge/test.csv.zip')
```

```python
classes = ["toxic", "severe_toxic", "obscene", "threat", "insult", "identity_hate"]
y = train_data[classes].values
sent_train = train_data["comment_text"]
sent_test = test_data["comment_text"]
```

```python
max_feature = 20000
tokenizer = Tokenizer(num_words=max_feature)
tokenizer.fit_on_texts(list(sent_train))
tokenized_train = tokenizer.texts_to_sequences(sent_train)
tokenized_test = tokenizer.texts_to_sequences(sent_test)
```

```python
maxlen = 100
X_train = pad_sequences(tokenized_train, maxlen=maxlen)
X_test = pad_sequences(tokenized_test, maxlen=maxlen)
```

```python
embed_size = 128
model = Sequential()
model.add(Embedding(max_feature, embed_size,input_length=maxlen))
model.add(LSTM(60, return_sequences=True,name='lstm_layer'))
model.add(GlobalMaxPool1D())
model.add(Dropout(0.1))
model.add(Dense(50, activation="relu"))
model.add(Dropout(0.1))
model.add(Dense(6, activation="sigmoid"))
model.compile(loss='binary_crossentropy',
                  optimizer='adam',
                  metrics=['accuracy'])
batch_size = 32
epochs = 2
model.fit(X_train,y, batch_size=batch_size, epochs=epochs, validation_split=0.1)
model.summary()
```

```python
out = model.predict(X_test)
out
```
