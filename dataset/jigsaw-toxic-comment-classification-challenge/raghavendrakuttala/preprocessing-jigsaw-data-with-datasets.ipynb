## Preprocessing using hugging face datasets library


```python
#since dataformat is in zip we can first load it into df then we import it into datasets format
train_df = pd.read_csv("/kaggle/input/jigsaw-toxic-comment-classification-challenge/train.csv.zip",
                      nrows=50)
```

### Loading data and basic metadata

1. import library
2. load data from dataframe
3. basic metadata functions like shape, num_columns, slicing etc..


```python
import datasets
```

```python
data = datasets.Dataset.from_pandas(train_df)
```

```python
data.shape #give shape as pandas
```

```python
data.num_columns #this function specifically to get num of columns
```

```python
data.column_names # to get column names
```

```python
data.features # to get feature and their data types
```

### cleaning data or using map functions

```python
import re
```

```python
def clean_text(row):
    row['comment_text'] = ''.join(re.findall(r'[\w ]*', row['comment_text']))
    return row
```

```python
# you can pass any function as long as it returns dict output as it take dict input
data.map(clean_text)[:3]
```
