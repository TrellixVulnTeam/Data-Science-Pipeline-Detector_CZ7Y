#### As you can observe in the video below, that belongs to the test set, there is a weird "shadow" moving in the background near the open door, as an erased person who enters in the room and soon later leaves the room...

```python
import os

from IPython.display import display, Audio, HTML #Video
from base64 import b64encode

def play_video(video_file, video_folder, pretitle=None, tipo='video/mp4'):
    '''
    Display video
    param: video_file - the name of the video file to display
    param: video_folder - the folder where the video file is located
    '''
    video_url = open(os.path.join(video_folder,video_file),'rb').read()
    data_url = "data:"+tipo+";base64," + b64encode(video_url).decode()
    title = video_file
    if pretitle is not None:
        title = pretitle+" "+title
    return HTML("""<h6>%s</h6><br><video width=500 controls><source src="%s" type="%s"></video>""" 
                % (title, data_url, tipo))
```
