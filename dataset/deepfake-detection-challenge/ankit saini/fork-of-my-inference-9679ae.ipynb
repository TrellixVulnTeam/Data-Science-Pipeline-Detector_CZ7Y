```python
# Don't change this
start = time.perf_counter()
n = 10 if speed_test else len(test_videos)
predictions = []
for i,video in enumerate(test_videos):
    if i==n:
        # if speed test
        break
        
    if i%20 ==19:
        # log at commit screen
        os.system(f'echo {str(i)} {predictions[-1]:.2f}')
    
    try:
        prediction = predict_on_video(test_dir+video)
        
        #test for valid prediction
        if prediction is not None: 
            predictions.append(prediction)
        else:
            predictions.append(0.5)
        
    except Exception as e:
        # Report Error at Commit 
        os.system(f'echo {str(e)}')
        predictions.append(0.5)

print(f'{(time.perf_counter()-start)/n:.2f}s per video. Should be less then 8s for a valid submission')
```

## Cleaning extra files created during prediction

```python
# Don't change this
import os, shutil
folder = '/kaggle/working'
for filename in os.listdir(folder):
    file_path = os.path.join(folder, filename)
    try:
        if os.path.isfile(file_path) or os.path.islink(file_path):
            os.unlink(file_path)
        elif os.path.isdir(file_path):
            shutil.rmtree(file_path)
    except Exception as e:
        print('Failed to delete %s. Reason: %s' % (file_path, e))
```
