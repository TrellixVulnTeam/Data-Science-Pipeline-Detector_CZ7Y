```python
import xgboost as xgb
import pandas as pd
import numpy as np
```

```python
train_df = pd.read_csv('../input/train.csv')
```

```python
# calculate levenshtein similarity for each pair of training questions
import Levenshtein as lev
tmplist = []
train_df['question1'] = train_df['question1'].astype(str)
train_df['question2'] = train_df['question2'].astype(str)
for i in train_df.index:
        tmplist.append(lev.ratio(train_df.get_value(i,'question1'),train_df.get_value(i,'question2')))
       
train_df.loc[:,'lev_sim'] = tmplist
```

```python
# find out the most frequent words used in questions then remove them
word_list = []
for i in train_df.index:
    for tmp in train_df.get_value(i,'question1').split():
        word_list.append(tmp)
                         
    for tmp in train_df.get_value(i,'question2').split():
        word_list.append(tmp)
```

```python
from collections import Counter
freq_counter= Counter(word_list)
print(freq_counter.most_common(1))
```
