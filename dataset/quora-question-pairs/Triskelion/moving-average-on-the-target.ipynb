# Moving average on the target

We use a moving average of window-size 10.000 over the target column `is_duplicate` and plot the mean for every window.

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def chunks(l, n):
    """Yield successive n-sized chunks from l."""
    for i in range(0, len(l), n):
        yield l[i:i + n]

df_train = pd.read_csv("../input/train.csv")
y = list(df_train["is_duplicate"])

means = []
for chunk in chunks(y, 10000):
    means.append(np.mean(chunk))

plt.plot(means)
```

```python
means = []
for chunk in chunks(y, 750):
    means.append(np.mean(chunk))

plt.plot(means)
```
