```python
import nltk
import string

from collections import Counter

from nltk.corpus import stopwords
```

```python
train_df = pd.read_csv("../input/train.csv")

train_df.head()
```

```python
def get_tokens(text):
    lowers = text.lower()
    #remove the punctuation using the character deletion step of translate
    no_punctuation = lowers.translate( string.punctuation)
    tokens = nltk.word_tokenize(no_punctuation)
    return tokens

train_df.question1 = train_df.question1.apply(lambda x: [w for w in get_tokens(x) if not w in stopwords.words('english')])
#train_df.question2 = train_df.question2.apply(lambda x: get_tokens(x))

train_df.head()
```
