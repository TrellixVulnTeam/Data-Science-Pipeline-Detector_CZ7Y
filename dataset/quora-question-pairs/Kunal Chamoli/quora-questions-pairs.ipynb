```python
df = pd.read_csv('../input/train.csv')
test = pd.read_csv('../input/test.csv')
```

```python
def extract_words(train):
    words = list()
    for index, row in train.iterrows():
        q1 = row['question1']
        q2 = row['question2']
        if not q1 or not q2 or not isinstance(q1, six.string_types) \
                or not isinstance(q2, six.string_types):
            continue
        q_words = q1.split()
        for word in q_words:
            words.append(word)
        q_words = q2.split()
        for word in q_words:
            words.append(word)
    return words
vocabulary_size = 50000
words = extract_words(df)
print('Number of words: %d' % len(words))
```
