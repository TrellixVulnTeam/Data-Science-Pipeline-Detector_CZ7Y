```python
import numpy as np 
import pandas as pd
from tqdm.auto import tqdm
import glob
import os
import cv2
```

```python
dim = (512, 512)
```

# Train data

```python
! mkdir train_images
```

```python
files = glob.glob("../input/sorghum-id-fgvc-9/train_images/*.png")
```

```python
for file in tqdm(files):
    image = cv2.imread(file)
    image = cv2.resize(image, dim, interpolation = cv2.INTER_AREA)
    name = os.path.basename(file)[:-3]
    cv2.imwrite(f'train_images/{name}jpg', image, [int(cv2.IMWRITE_JPEG_QUALITY), 100])
```
