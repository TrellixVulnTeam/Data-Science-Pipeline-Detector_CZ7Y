```python
import pandas as pd
train=pd.read_csv('../input/sorghum-cultivar-identification-128128/train.csv')
train.head()
```

```python
cultivar = train.sort_values('cultivar')
cultivar.head()
```

```python
# take unique labels
class_names = list(cultivar.cultivar.unique())
class_names[:5]
```

```python
import shutil, os
newpath = r'./train' 
if not os.path.exists(newpath):
    os.makedirs(newpath)
```

```python
#creating subfolders
for c in class_names[1:]:
    dest =  r'./train/'+str(c)
    os.makedirs(dest)
    for i in list(cultivar[cultivar['cultivar']==c]['image']): # Image Id
        get_image = os.path.join('../input/sorghum-cultivar-identification-128128/train/', i) # Path to Images 
        try:
            move_image_to_cat = shutil.copy(get_image, dest)
        except OSError as e:
            continue
   
```

```python
newpath = r'./test' 
if not os.path.exists(newpath):
    os.makedirs(newpath)
```

```python
import glob
src_dir = "../input/sorghum-cultivar-identification-128128/test"
dst_dir = "./test"
for jpgfile in glob.iglob(os.path.join(src_dir, "*.png")):
    shutil.copy(jpgfile, dst_dir)
```
