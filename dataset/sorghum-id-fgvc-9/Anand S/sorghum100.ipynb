# Visualise

```python
# plotting image by image id for single image
def  visualize(image_id):
    
    path = df_train.loc[df_train['image'] == image_id, 'image_path'].iloc[0]
    cultivar = df_train.loc[df_train['image'] == image_id, 'cultivar'].iloc[0]
    
    plt.figure(figsize=(10, 10))
    image = cv2.imread(path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image)
    plt.title(f"CULTIVAR: {cultivar} \n Image:{image_id}", fontsize=10)
    plt.show()
```

```python
visualize('2017-06-16__12-24-20-930.png')
```

```python
# plotting image by image id for multiple image. When an image id is given, function returns 20 images 
#in same class
def visualize_many(image_id):
    train_image_path = "../input/sorghum-id-fgvc-9/train_images/"
    
    cultivar = df_train.loc[df_train['image'] == image_id, 'cultivar'].iloc[0]
    df = df_train.loc[df_train['cultivar'] == cultivar]
    
    
                                                           
    if  df['image'].count()< 20:
        x=df['image'].tolist()
        plt.figure(figsize=(18, 18))
        for i, j in zip(x, range(20)):       
            plt.subplot(5, 4, j + 1)
            path = df.loc[df['image'] == i, 'image'].iloc[0]
            image = cv2.imread(os.path.join(train_image_path, path))
            image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
            plt.imshow(image)
            cultivar = df.loc[df['image'] == i, 'cultivar'].iloc[0]
            imageid= df.loc[df['image'] == i, 'image'].iloc[0]
            plt.title(f"CULTIVAR: {cultivar} \n Image:{imageid}", fontsize=10)
            plt.axis("off")
        plt.show()
    else:
        x = np.random.choice(df['image'], 20, replace=False).tolist()
        plt.figure(figsize=(18, 18))
        for i, j in zip(x, range(20)):       
            plt.subplot(5, 4, j + 1)
            path = df.loc[df['image'] == i, 'image'].iloc[0]
            image = cv2.imread(os.path.join(train_image_path, path))
            image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
            plt.imshow(image)
            cultivar = df.loc[df['image'] == i, 'cultivar'].iloc[0]
            imageid= df.loc[df['image'] == i, 'image'].iloc[0]
            plt.title(f"CULTIVAR: {cultivar} \n Image:{imageid}", fontsize=10)
            plt.axis("off")
    
        plt.show()
```

```python
visualize_many('2017-06-16__12-24-20-930.png')
```

```python
#plot image by cultivar name
def visualize_cultivar(name):
    train_image_path = "../input/sorghum-id-fgvc-9/train_images/"
     
    df = df_train.loc[df_train['cultivar'] == name]
    
    x = np.random.choice(df['image'], 5, replace=False).tolist()
    plt.figure(figsize=(18, 18))
    for i, j in zip(x, range(15)):       
            plt.subplot(1, 5, j + 1)
            path = df.loc[df['image'] == i, 'image'].iloc[0]
            image = cv2.imread(os.path.join(train_image_path, path))
            image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
            plt.imshow(image)
            cultivar = df.loc[df['image'] == i, 'cultivar'].iloc[0]
            imageid= df.loc[df['image'] == i, 'image'].iloc[0]
            plt.title(f"CULTIVAR: {cultivar} \n Image:{imageid}", fontsize=10)
            plt.axis("off")
            #plt.savefig('saved_figure.png')
    
    plt.show()
```
