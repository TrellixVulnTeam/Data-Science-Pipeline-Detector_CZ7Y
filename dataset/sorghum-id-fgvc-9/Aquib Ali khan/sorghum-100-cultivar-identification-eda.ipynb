```python
import os
import cv2
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
BASE_DIR = "../input/sorghum-id-fgvc-9/train_images"
df_train = pd.read_csv("../input/sorghum-id-fgvc-9/train_cultivar_mapping.csv")
```

```python
plt.figure(figsize=(8, 25))
sns.countplot(y="cultivar", data=df_train,);
```

<h2><center>Just a visualization<center><h2/>

```python
def visualize_batch(image_ids, labels):
    plt.figure(figsize=(16, 12))
    
    for ind, (image_id, label) in enumerate(zip(image_ids, labels)):
        plt.subplot(4, 4, ind + 1)
        image = cv2.imread(os.path.join(BASE_DIR, image_id))
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        plt.imshow(image)
        plt.title(f"Class: {label}", fontsize=12)
        plt.axis("off")
    
    plt.show()
```

```python
tmp_df = df_train.sample(12)
image_ids = tmp_df["image"].values
labels = tmp_df["cultivar"].values

visualize_batch(image_ids, labels)
```

```python
label = df_train.cultivar.dropna().unique()
for i in label:
    
    tmp_df = df_train[df_train["cultivar"] == str(i)]
    print(f"Total train images for class {i}: {tmp_df.shape[0]}")

    tmp_df = tmp_df.sample(4)
    image_ids = tmp_df["image"].values
    labels = tmp_df["cultivar"].values

    visualize_batch(image_ids, labels)
```
