```python
train2=train.drop(['datetime', 'casual', 'registered', 'count'], axis=1)
train2
```

```python
test2=test.drop(['datetime'], axis=1)
test2
```

```python
from sklearn.model_selection import train_test_split
x_train, x_valid, y_train, y_valid = train_test_split(train2, train['registered'], random_state=42)
x_train.shape, x_valid.shape, y_train.shape, y_valid.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(n_jobs=4, random_state=42)
rf.fit(x_train, y_train) 
```

```python
result=rf.predict(test2)
result
```
