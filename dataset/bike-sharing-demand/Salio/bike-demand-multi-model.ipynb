## Train your model (Regression)

```python
## Simple Linear Regression (SLR) by registered (Strong relationship )
X = Bike_T[['registered']]
y = Bike_T['count']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size =0.25, random_state = 4)
```

```python
# Import library for Linear Regression
from sklearn.linear_model import LinearRegression

# Create a Linear regressor
Simple_LR = LinearRegression()

# Train the model using the training sets 

Simple_LR.fit(X_train, y_train)
```

```python
# y = ax+b this is            a: y= 1.1x +b
Simple_LR.coef_ 
```

```python
# y= ax + b :                 y =1.1x+10.44
Simple_LR.intercept_
```

```python
y_pred=Simple_LR.predict(X_test)
```

```python
X_test
```

## Model (SLR) Evaluation

```python
y_test
```

```python
y_pred
```

```python
sns.histplot(y_test-y_pred)
```

```python
from sklearn import metrics
print('R^2:',metrics.r2_score(y_test, y_pred))
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
# Visualizing the differences between actual count and predicted count
plt.scatter(y_test, y_pred , color = 'r')
plt.xlabel("count")
plt.ylabel("Predicted count")
plt.title("Prices vs Predicted count")
plt.show()
```

## Multiple Linear Regression (MLR)

```python
Bike_T = pd.read_csv('../input/bike-sharing-demand/train.csv')
X = Bike_T.drop(["count","datetime","casual","registered"],axis = 1)
y = Bike_T["count"]
type(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.25, random_state = 4)
```

```python
from sklearn.linear_model import LinearRegression
Multi_LR = LinearRegression()
Multi_LR.fit(X_train, y_train)
```

```python
Multi_LR.intercept_
```

```python
Multi_LR.coef_
```

```python
coeffcients = pd.DataFrame([X_train.columns,Multi_LR.coef_]).T
coeffcients = coeffcients.rename(columns={0: 'Attribute', 1: 'Coefficients'})
coeffcients
```

## Model(MLR) Evaluation

```python
y_pred = Multi_LR.predict(X_test)
```

```python
print("R^2 : "  ,metrics.r2_score(y_test,y_pred))
print("MSE  :"  ,metrics.mean_squared_error(y_test,y_pred))
print("MAE : "  ,metrics.mean_absolute_error(y_test,y_pred))
print("RMSE : ", np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

```python
# Visualizing the differences between actual acount and predicted acount
plt.scatter(y_test, y_pred)
plt.xlabel("acount")
plt.ylabel("Predicted acount")
plt.title("acount vs Predicted acount")
plt.show()
```

```python
sns.displot(y_test-y_pred)
plt.title("Histogram of Residuals")
plt.xlabel("Residuals")
plt.ylabel("Frequency")
plt.show()
```

# Polynomial regression 

```python
Bike_T = pd.read_csv('../input/bike-sharing-demand/train.csv')
X = Bike_T.drop(["count","datetime","casual","registered"],axis = 1)
y = Bike_T["count"]
type(X)
type(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.25, random_state = 4)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
"Creates a polynomial regression model for the given degree"
poly_features = PolynomialFeatures(degree=2)
   
# transform the features to higher degree features.
X_train_biquadratic = poly_features.fit_transform(X_train)
   
# fit the transformed features to Linear Regression
biquadratic = LinearRegression()

biquadratic.fit(X_train_biquadratic, y_train)
     
# predicting on training data-set
y_train_predicted = biquadratic.predict(X_train_biquadratic)
   
# predicting on test data-set
y_test_predicted = biquadratic.predict(poly_features.fit_transform(X_test))
print(y_test_predicted)
```

```python
metrics.r2_score(y_train, y_train_predicted)
```

```python
metrics.r2_score(y_test, y_test_predicted)
```

```python
# Visualizing the differences between actual acount and predicted acount
plt.scatter(y_test, y_test_predicted)
plt.xlabel("acount")
plt.ylabel("Predicted acount")
plt.title("acount vs Predicted acount")
plt.show()
```

```python
sns.displot(y_test-y_test_predicted)
plt.title("Histogram of Residuals")
plt.xlabel("Residuals")
plt.ylabel("Frequency")
plt.show()
```
