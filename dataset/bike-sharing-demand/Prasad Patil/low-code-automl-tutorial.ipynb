### Performing Simple AutoML Task with [PyCaret Library](https://pycaret.org)
PyCaret is low code fully open source automatic machine learning library which helps from <br>
data preparation to final model deployment in every mode of learning i.e Supervised, Unsupervised <br> or Clustering.

Importing libraries and dataset 

```python
from pycaret.regression import *
import numpy as np 
import pandas as pd
train = pd.read_csv('/kaggle/input/bike-sharing-demand/train.csv')
train.head()
```

```python
test = pd.read_csv('/kaggle/input/bike-sharing-demand/test.csv')
test.head()
```

As 'casual' and 'registered' are extra columns which are not present in test set so droping them 

```python
features_drop = ['casual','registered']
train.drop(features_drop,axis = 1,inplace = True)
```

Finding null columns though PyCaret manages to fill null values

```python
print(train.isnull().sum(axis=0))
print(test.isnull().sum(axis=0))
```

Loading and normalizing the dataset into Pycaret 

```python
stp = setup(train, target="count", normalize = True)
```

From above cell the number of featuers PyCaret is providing for just feature engineering can be seen.

Rather fitting dataset individually to every ML algorithm, Pycaret gives feature to compare between them directly.

```python
models = compare_models(n_select=3)    # n_select - selecting top 3 best performed models
```

Comparing models with each other can be the best possible feature of AutoML where time efficiency can be obtained. <br>After this hyperparameter tuning can also be done to that particular best performed model or even of top selects, this all <br>may a downhill if there is lack of high end GPU support.
