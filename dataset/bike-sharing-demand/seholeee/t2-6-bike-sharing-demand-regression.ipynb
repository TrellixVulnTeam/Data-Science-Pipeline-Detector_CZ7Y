```python
# 시험환경 세팅

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

def exam_data_load(data, target, id_name = '', null_name = ''):
    if id_name == '':
        data = data.reset_index().rename(columns = {'index' : 'id'})
        id_name = 'id'
    else:
        id_name = id_name
        
    if null_name != '':
        data[data == null_name] = np.nan
        
    X_train, X_test = train_test_split(data, test_size = 0.2, shuffle = True, random_state = 2021)
    y_train = X_train[[id_name, target]]
    X_train = X_train.drop(columns = [id_name, target])
    y_test = X_test[[id_name, target]]
    X_test = X_test.drop(columns = [id_name, target])
    return X_train, X_test, y_train, y_test

data = pd.read_csv('../input/bike-sharing-demand/train.csv')
X_train, X_test, y_train, y_test = exam_data_load(data, target = 'count')

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
X_train.head()
```

```python
X_train['datetime'] = pd.to_datetime(X_train['datetime'])
X_test['datetime'] = pd.to_datetime(X_test['datetime'])
```

```python
X_train['year'] = X_train['datetime'].dt.year
X_train['month'] = X_train['datetime'].dt.month
X_train['day'] = X_train['datetime'].dt.day
X_train['hour'] = X_train['datetime'].dt.hour

X_test['year'] = X_test['datetime'].dt.year
X_test['month'] = X_test['datetime'].dt.month
X_test['day'] = X_test['datetime'].dt.day
X_test['hour'] = X_test['datetime'].dt.hour
```

```python
X_train = X_train.drop(['datetime', 'casual', 'registered'], axis = 1)
X_test = X_test.drop(['datetime', 'casual', 'registered'], axis = 1)

y_train = y_train.drop('id', axis = 1)
```

```python
X_tr, X_val, y_tr, y_val = train_test_split(X_train, y_train, test_size = 0.3, random_state = 42)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score

rf = RandomForestRegressor(random_state = 42)
rf.fit(X_tr, y_tr)
pred_val = rf.predict(X_val)
print(r2_score(y_val, pred_val))
```

```python
rf.fit(X_train, y_train)
pred = rf.predict(X_test)
print(r2_score(y_test['count'], pred))
submission = pd.DataFrame(pred)
submission.to_csv('00000.csv', index = False)
```
