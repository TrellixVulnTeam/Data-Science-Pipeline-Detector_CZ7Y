working on making a new variable showing the type of the day whather this day is a holiday -> 0 or weekend -> 1 or working day -> 2 .... "meging the two cols working day and holiday"

```python
def day_type(row):
    if row['holiday'] == 1:
        return 0
    elif row['workingday'] == 0:
        return 1
    else:
        return 2
# 0 -> holiday, 1 -> weekend, 2 -> workingday
```

```python
train_df['day_type'] = train_df.apply(lambda row : day_type(row), axis=1)
#train_df['day_type'].value_counts()
```

for test_df

```python
test_df['day_type'] = test_df.apply(lambda row : day_type(row), axis=1)
test_df['day_type'].value_counts()
```

we can also make a new variable to see that day is a week end or not as this is a very important insight and has a strong correlation with the bike demand

```python
def weekend(row):
    if row['day_type'] == 1:
        return 1
    else:
        return 0
# 1 -> weekend, 1 -> notweekend
```

```python
train_df['weekend'] = train_df.apply(lambda row : weekend(row), axis=1)
train_df['weekend'].value_counts()
```

for test_df

```python
test_df['weekend'] = test_df.apply(lambda row : weekend(row), axis=1)
test_df['weekend'].value_counts()
```

we have seen from data visualization that with time the bike demand is increasing so lets make a new discrete variable "quarters" with its increas the count will increase

```python
def quarter(row):
    if row['years'] == 2011:
        if row['months'] < 4:
            return 0
        elif row['months'] < 7:
            return 1
        elif row['months'] < 10:
            return 2
        else:
            return 3
    else:
        if row['months'] < 4:
            return 4
        elif row['months'] < 7:
            return 5
        elif row['months'] < 10:
            return 6
        else:
            return 7
# 0 -> 1st Quart, 1 -> 2nd Quart, .... , 7 -> 8th Quart
```

```python
train_df['quarter'] = train_df.apply(lambda row : quarter(row), axis=1)
#train_df['quarter'].value_counts()
```

for test_df

```python
test_df['quarter'] = test_df.apply(lambda row : quarter(row), axis=1)
#test_df['quarter'].value_counts()
```

# Preparing Data for the Model

at the model we will predict two outputs registered and casual and then we will apply the summation to calculate the count

```python
X_cas = train_df.drop(['datetime', 'casual', 'registered', 'count', 'hours_intervals_reg', 'temp_intervals_reg'], axis=1)
Y_cas = np.log(train_df['casual']+1) # we take the log as to avoid the effect of the outliers

X_reg = train_df.drop(['datetime', 'casual', 'registered', 'count', 'hours_intervals_cas', 'temp_intervals_cas'], axis=1)
Y_reg = np.log(train_df['registered']+1)
```

for test_df

```python
X_cas_test = test_df.drop(['datetime', 'hours_intervals_reg', 'temp_intervals_reg'], axis=1)

X_reg_test = test_df.drop(['datetime', 'hours_intervals_cas', 'temp_intervals_cas'], axis=1)
```

# Random Forest Model

```python
from sklearn.model_selection import cross_validate
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import make_regression
from sklearn.metrics import SCORERS
from sklearn.metrics import make_scorer
from sklearn.model_selection import GridSearchCV
```

```python
# For casual data
random_state = 42
rand_Fors_cas = RandomForestRegressor(random_state = random_state)
param_grid = {"max_depth" : [50]}

grid_cas = GridSearchCV(estimator = rand_Fors_cas, param_grid=param_grid, scoring = 'neg_root_mean_squared_error',
                    verbose = 1, n_jobs = -1, cv = 10)
grid_cas.fit(X_cas,Y_cas)
print("Best Score for casual:" + str(grid_cas.best_score_))
print("Best Parameters for casual: " + str(grid_cas.best_params_))
best_parameters_cas = grid_cas.best_params_
```

```python
rand_Fors_cas = RandomForestRegressor(random_state = random_state,**best_parameters_cas)
rand_Fors_cas.fit(X_cas,Y_cas)
```

```python
#For registered data
random_state = 42
rand_Fors_reg = RandomForestRegressor(random_state = random_state)
param_grid = {"max_depth" : [50]}

grid_reg = GridSearchCV(estimator = rand_Fors_reg, param_grid=param_grid, scoring = 'neg_root_mean_squared_error',
                    verbose = 1, n_jobs = -1, cv = 10)
grid_reg.fit(X_reg,Y_reg)
print("Best Score for casual:" + str(grid_reg.best_score_))
print("Best Parameters for casual: " + str(grid_reg.best_params_))
best_parameters_reg = grid_cas.best_params_
```

```python
rand_Fors_reg = RandomForestRegressor(random_state = random_state,**best_parameters_reg)
rand_Fors_reg.fit(X_reg,Y_reg)
```

# Test Prediction

```python
# For casual data
test_df['log_y_casual_test'] = rand_Fors_cas.predict(X_cas_test)
test_df['y_casual_test'] = np.exp(test_df['log_y_casual_test'])

#For registered data
test_df['log_y_registered_test'] = rand_Fors_reg.predict(X_reg_test)
test_df['y_registered_test'] = np.exp(test_df['log_y_registered_test'])
```

```python
# Calculate Count
test_df['count'] = test_df['y_casual_test'] + test_df['y_registered_test']
test_df['count'] = test_df['count'].round()
```
