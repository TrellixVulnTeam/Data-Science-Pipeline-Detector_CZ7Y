```python
alldata = pd.concat([train, test])
alldata
```

```python
alldata_corr = alldata.corr()['count'].sort_values()
alldata_corr
```

```python
alldata['datetime'] = pd.to_datetime(alldata['datetime'])

alldata['hour'] = alldata['datetime'].dt.hour
alldata['year'] = alldata['datetime'].dt.year
alldata['weekday'] = alldata['datetime'].dt.weekday
```

```python
alldata_corr = alldata.corr()['count'].sort_values()
alldata_corr
```

```python
workingday_df = alldata[alldata['workingday'] == 1]
weekend_df = alldata[alldata['workingday'] ==0]
```
