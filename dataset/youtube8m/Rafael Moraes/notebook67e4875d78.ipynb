This is a very interesting competition that aims to explore the YouTube-8M data set, which is a recent creation by Google (released on Sep/2016) and has millions of YouTube videos previously analyzed and classified. Its objective is to be a reliable data set for data scientists to experiment new algorithms for video classification. The YouTube-8M data set is to videos what the ImageNet data set is to images.

So, let's start by clarifying some points:

 - The YouTube-8M data set is enormous, containing more than 8 million videos previously labeled (more than one label per video is possible).
 - It would be impossible for us, mere mortals, to analyze all the frames from all the videos, so the research team at Google made it easy for us by using pre-trained neural networks (NNs) to extract relevant features from the videos.
 - Only the first 6 minutes of each video were analyzed by these NNs, one frame per second (so 360 frames per video). And as a result, from each frame, **1024 video features** and **128 audio features** were extracted. These are the **frame-level features**.
 - As a simplified alternative, these frame-level features are averaged (also centered and decorrelated) for all the analyzed frames in a video and the result is the **video-level features**.
