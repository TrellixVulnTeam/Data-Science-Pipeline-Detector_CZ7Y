```python
events = pd.read_csv(os.path.join(directory, 'events.csv'))
```

```python
events['timestamp'] = pd.to_datetime(events['timestamp'])
type(events['timestamp'])
```

```python
events['day_of_week'] = events['timestamp'].dt.weekday_name #dayofweek
events.head()
```

```python
events['timestamp'] = pd.to_datetime(events['timestamp'])

events['month'] = events['timestamp'].dt.month #dayofweek
np.unique(events['month'])
```

```python
type(events['timestamp'])
```

```python

events['hour'] = events['timestamp'].dt.hour
events.head()
```

```python
freq = pd.DataFrame(events.groupby(["device_id"]).size().reset_index(name="total_device_active_event"))
freq.head() # active_count is no of distinct event each device is active on
```

```python
events = pd.merge(events, freq, on = 'device_id', how = 'left' )
```
