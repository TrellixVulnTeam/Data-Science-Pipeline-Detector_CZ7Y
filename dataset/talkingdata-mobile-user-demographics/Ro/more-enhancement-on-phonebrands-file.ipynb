```python
phone_brand_data = pd.read_csv('../input/phone_brand_device_model.csv')
phone_brand_data['brand'] = phone_brand_data.phone_brand.map(brands_map)
phone_brand_data.to_csv('new_phone_brands', columns=['device_id', 'brand'], index=False)
```

```python
null_brands =  phone_brand_data[phone_brand_data.brand.isnull()][['device_id', 'phone_brand']].groupby(['phone_brand']) \
                .count()

null_brands.reset_index(level=0, inplace=True)
null_brands = null_brands.rename(columns={'device_id' : 'count'})

print(null_brands.sort_values('count', ascending=False))

print ("\n Unknown Phone Brands - " + str(null_brands['count'].sum()))
```
