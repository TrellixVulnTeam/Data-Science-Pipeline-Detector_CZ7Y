[H2O AutoML](http://docs.h2o.ai/h2o/latest-stable/h2o-docs/automl.html) is an automated machine learning meta-algorithm that is part of the [H2O software library](http://docs.h2o.ai/h2o/latest-stable/h2o-py/docs/intro.html#what-is-h2o). (It shold not be confused with [H2O DriverlessAI](https://www.h2o.ai/products/h2o-driverless-ai/), which is a commercial product and built from an entirely different code base.) H2O’s AutoML can be used for automating the machine learning workflow, which includes automatic training and tuning of many models within a user-specified time-limit. Stacked Ensembles – one based on all previously trained models, another one on the best model of each family – will be automatically trained on collections of individual models to produce highly predictive ensemble models which, in most cases, will be the top performing models in the AutoML Leaderboard.# This Python 3 environment comes with many helpful analytics libraries installed

```python
import h2o
print(h2o.__version__)
from h2o.automl import H2OAutoML

h2o.init(max_mem_size='16G')
```

```python
train = h2o.import_file("../input/baseline-feature-engineering-geotab-69-5-lb/train_danFeatsV1.csv.gz")
test = h2o.import_file("../input/baseline-feature-engineering-geotab-69-5-lb/test_danFeatsV1.csv.gz")
```

```python
x = ["IntersectionId",
             'Intersection',
           'diffHeading',  'same_street_exact',
           "Hour","Weekend","Month",
          'Latitude', 'Longitude',
          'EntryHeading', 'ExitHeading',
            'Atlanta', 'Boston', 'Chicago',
       'Philadelphia']

y = 'DistanceToFirstStop_p20'
```

```python
aml = H2OAutoML(max_models=1000, seed=121, max_runtime_secs=30000)
aml.train(x=x, y=y, training_frame=train)
```

```python
# View the AutoML Leaderboard
lb = aml.leaderboard
lb.head(rows=lb.nrows)  # Print all rows instead of default (10 rows)
```

```python
# The leader model is stored here
aml.leader
```

```python
# If you need to generate predictions on a test set, you can make
# predictions directly on the `"H2OAutoML"` object, or on the leader
# model object directly

preds = aml.predict(test)
```

```python
preds = preds.as_data_frame().values.flatten()
preds
```

```python
np.save('preds_'+y, preds)
```
