You may have noticed that the tif files look very funky in a notebook. This happens for a number of reasons, including the fact that these are 16-bit multi-spectral images, the band order is not what you'd expect  (blue, green, red, near IR), and the fact that the cameras creating the images create 12-bit images. Fortunately, the [`spectral`](http://www.spectralpython.net/) library has some convenience functions to make this easier.

See below how `plt.imshow` and `imshow` (from `spectral`) look quite different.
