```python
df = pd.read_csv('../input/train.csv')
df.head()
```

```python
def extract_labels(df):
    label_list = df['tags'].apply(lambda x: x.split(' '))
    ret = set()
    
    for labels in label_list:
        for l in labels:
            ret.add(l)
    return list(ret)

labels = extract_labels(df)
for l in labels:
    df[l] = 0
        
```

```python
# Add all each label as a row to the dataframe


for i,v in df.iterrows():
    labels = df.loc[i]['tags'].split(' ')
    for label in labels:
        df.loc[i][label] = 1
    
 

df.head()
```
