```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
import os
import sys
import itertools

#read data
training_data = pd.read_csv('../input/train.csv')
training_data['tags'] = training_data['tags'].astype('str') 
```

```python
#get all unique tags
tags = np.unique(np.array(list(itertools.chain(*[x.split(' ') for x in training_data['tags'].values]))))
```

```python
# create a record --> tags mapping

tags_of_training_data = np.zeros((len(training_data),len(tags)))
tags_of_training_data.shape
```

```python
# create a record --> tags mapping
tags_of_training_data = np.zeros((len(training_data),len(tags)),dtype=int)

tag_frequency = np.zeros(len(tags))

for i, tag in enumerate(tags):
    locations = np.where(np.char.find(
        training_data['tags'].values.astype(str),tag)>-1)[0]
    tags_of_training_data[locations,i] = 1
    tag_frequency[i] = len(locations)

print('Example mappings from record to tags')
print(tags_of_training_data[:10,:])
```

```python
tag_info = pd.DataFrame(data={'tag':tags, 'frequency':tag_frequency})
```

```python
tag_indices = np.arange(len(tags))
fig, ax = plt.subplots()
plt.barh(tag_indices, list(tag_frequency))
ax.set_yticks(tag_indices)
ax.set_yticklabels(tags)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('# Images')
plt.show()
```

```python
sns.barplot(data=tag_info, x='tags',y='frequency')
```
