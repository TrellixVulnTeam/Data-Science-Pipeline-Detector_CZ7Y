```python
import tifffile as tiff
from scipy import ndimage
a = tiff.imread('../input/train-tif/train_12107.tif')
#print(a.shape)
#print(a.dtype)
npa = np.array(a)
import matplotlib.pyplot as plt
#plt.imshow(a[:,:,1])
#plt.show()

# extract the different bands
b, g, r, nir = npa[:, :, 0], npa[:, :, 1], npa[:, :, 2], npa[:, :, 3]
# Allow division by zero
np.seterr(divide='ignore', invalid='ignore')
# compute ndvi
ndvi = (nir - r)/(nir + r)
# plot all bands and ndvi
fig = plt.figure()
fig.set_size_inches(12, 4)
for i, (x, c) in enumerate(((r, 'r'), (g, 'g'), (b, 'b'), (nir, 'near-ir'), (ndvi, 'ndvi'))):
    subplt = fig.add_subplot(1, 5, i+1)
    subplt.set_title(c)
    plt.imshow(x, cmap=plt.cm.gray)
plt.show()
```
