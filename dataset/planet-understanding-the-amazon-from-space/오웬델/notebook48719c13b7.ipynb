```python
from skimage import io

image = io.imread("../input/train-tif/train_0.tif")
red = image[:,:,0]
```

```python
redFlat = red.flatten()
redRange = max(redFlat) - min(redFlat)
redRange
```

```python
import matplotlib.pyplot as plt
hist = plt.hist(redFlat, bins=redRange)
```

```python

powers = list(map(lambda x: 2**x, range(16)))
count = {}
for power in powers:
    count[power] = 0
for value in red.flatten():
    for power in powers:
        if value & power != 0:
            count[power] += 1
count
```
