```python
lbl = 'primary'
train_labels[lbl] = 1 * train_labels['tags'].apply(lambda row: lbl in row.split(" "))
print(train_labels[lbl].sum())


lbl = 'clear'
train_labels[lbl] = 1 * train_labels['tags'].apply(lambda row: lbl in row.split(" "))
print(train_labels[lbl].sum())


lbl = 'agriculture'
train_labels[lbl] = 1 * train_labels['tags'].apply(lambda row: lbl in row.split(" "))
print(train_labels[lbl].sum())


lbl = 'road'
train_labels[lbl] = 1 * train_labels['tags'].apply(lambda row: lbl in row.split(" "))
print(train_labels[lbl].sum())


train_labels.head(3)
```

```python
from glob import glob

image_paths = sorted(glob('../input/train-jpg/*.jpg'))
print(len(image_paths))

image_names = list(map(lambda row: row.split("/")[-1][:-4], image_paths))
image_names[0:10]
```

```python
for img in image_paths[:5]:
    print(img)
```

```python
from PIL import Image
import matplotlib.pyplot as plt

im = plt.imread(image_paths[0])
#im = Image.open(image_paths[0])
#im = np.array(im)[:,:,:3]

print(train_labels[train_labels.image_name==image_names[0]])
print(im.shape)
print(im[:,:,3].mean(), im[:,:,3].std())

plt.imshow(im)
```

```python
train_labels['r_avg'] = np.nan
train_labels['g_avg'] = np.nan
train_labels['b_avg'] = np.nan
train_labels['a_avg'] = np.nan
```

```python
import time

t0 = time.time()

for e, img_path in enumerate(image_paths[:15000]):
    
    # Load image
    im = plt.imread(img_path)
    
    # Set values
    ix = train_labels.image_name==image_names[e]
    train_labels.loc[ix, 'r_avg'] = im[:,:,0].mean()
    train_labels.loc[ix, 'g_avg'] = im[:,:,1].mean()
    train_labels.loc[ix, 'b_avg'] = im[:,:,2].mean()
    train_labels.loc[ix, 'a_avg'] = im[:,:,3].mean()


print(time.time()-t0)

train_labels.head(3)
```

```python
print(train_labels[train_labels.a_avg>0.0].primary.sum()/10000)
print(train_labels.primary.sum()/train_labels.shape[0])
```

```python
from sklearn.linear_model import LogisticRegression

X = train_labels[train_labels.a_avg>0.0][['r_avg','g_avg','b_avg']].values
y = train_labels[train_labels.a_avg>0.0].primary.values

reg = LogisticRegression(penalty='l2', solver='lbfgs', max_iter=100, random_state=42)
reg.fit(X, y)

from sklearn.metrics import fbeta_score, confusion_matrix

y_pred = reg.predict(X)

print((y==y_pred).sum())
print((y==y_pred).sum()/y.shape[0])

confusion_matrix(y, y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier

X = train_labels[train_labels.a_avg>0.0][['r_avg','g_avg','b_avg']].values
y = train_labels[train_labels.a_avg>0.0].primary.values

clf = RandomForestClassifier(n_estimators=50)
clf.fit(X, y)

y_pred_rf = clf.predict(X)

print((y==y_pred_rf).sum())
print((y==y_pred_rf).sum()/y.shape[0])

confusion_matrix(y, y_pred_rf)
```
