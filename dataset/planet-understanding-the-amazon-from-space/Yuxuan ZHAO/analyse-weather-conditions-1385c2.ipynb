```python
df = pd.read_csv("../input/train_v2.csv")
tags = df["tags"].apply(lambda x: x.split(' '))
   
end = len(tags)
id_haze = []
id_cloudy = []
id_partly = []
id_clear = []

for i in range (0,end):
    for x in tags[i]:
        if x == 'haze':
            id_haze.append(i)
        elif x == 'cloudy':
            id_cloudy.append(i)
        elif x == 'partly_cloudy':
            id_partly.append(i)
        elif x == 'clear':
            id_clear.append(i)
print (len(id_haze))
print (len(id_cloudy))
print(len(id_partly))
print (len(id_clear))
```

```python
import cv2
import matplotlib.pyplot as plt
import random

new_style = {'grid': True}
plt.rc('axes', **new_style)
_, ax = plt.subplots(3, 3, sharex='col', sharey='row', figsize=(12, 12))
index = []
for i in range(0,9):
 
    if i <3:
        l = random.choice(id_cloudy)
        index.append(l)
    elif (i>=3 and i<6):
        l = random.choice(id_partly)
        index.append(l)
    elif (i>=6 and i<9):
        l = random.choice(id_haze)
        index.append(l)
    
    img = cv2.imread('../input/train-jpg/train_'+str(l)+'.jpg')
    ax[i // 3, i % 3].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

plt.show()
print (index)
```

```python
# feature extractor 
from skimage.feature import local_binary_pattern
from skimage.feature import hog 
H = []
S = []
V = []
HOG = []
LBP = []
P = 8
R = 4
L = len (fenetre)
for i in range(0,L):
    window = cv2.cvtColor(fenetre[i], cv2.COLOR_BGR2HSV)
    H.append (window[:,:,0])
    S.append ( window[:,:,1])
    V.append (window[:,:,2])

for i in range(0,L):
    window = cv2.cvtColor(fenetre[i], cv2.COLOR_BGR2GRAY)
    lbp = local_binary_pattern(window,P,R)
    hog_ft = hog(window, orientations=9, pixels_per_cell=(4, 4), cells_per_block=(3, 3))
    LBP.append(lbp)
    HOG.append(hog_ft)
         
```
