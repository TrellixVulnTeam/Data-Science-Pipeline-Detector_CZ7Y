```python
import numpy as np 
import pandas as pd
from spectral import *
from skimage import io
import os
from sklearn.preprocessing import MinMaxScaler
from subprocess import check_output
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
print(check_output(["ls", "../input"]).decode("utf8"))
df = pd.read_csv('../input/train_v2.csv')
np.seterr(all='warn') # divide by zero, NaN values
# Any results you write to the current directory are saved as output.
```

```python
basepath = '../input/train-tif-v2'
train_img = 'train_131'
print(df.loc[df['image_name']==train_img, 'tags'].to_string())
```

```python
tif_image = io.imread(os.path.join(basepath, '{}.tif'.format(train_img)))
```

```python
def scale(img):
    rescaleIMG = np.reshape(img, (-1, 1))
    scaler = MinMaxScaler(feature_range=(0, 255))
    rescaleIMG = scaler.fit_transform(rescaleIMG) # .astype(np.float32)
    img_scaled = (np.reshape(rescaleIMG, img.shape)).astype(np.uint8)
    return img_scaled
```

```python
def calculate_ndvi(image):
    """
    Calculate normalized difference vegetation index as follows: (NIR-R)/(NIR+R)
    Input image is in the format (NIR, R, G)
    """
    return (image[:, :, 0] - image[:, :, 1]) / (image[:, :, 0] + image[:, :, 1])

def calculate_nwvi(image):
    """
    Calculate normalized difference water index as follows: (G-NIR)/(G+NIR)
    Input image is in the format (NIR, R, G)
    """
    return (image[:, :, 2] - image[:, :, 0])/(image[:, :, 2] + image[:, :, 0])
```

```python
img2 = scale(get_rgb(tif_image, [2, 1, 0])) # RGB
img3 = scale(get_rgb(tif_image, [3, 2, 1])) # RG NIR
img4 = scale(get_rgb(tif_image, [3])) # NIR
water = calculate_ndvi(img3)
veg = calculate_nwvi(img3)
```
