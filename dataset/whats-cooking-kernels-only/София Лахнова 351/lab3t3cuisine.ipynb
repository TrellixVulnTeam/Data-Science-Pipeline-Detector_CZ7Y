```python
#сделали из списка продуктов, текст состоящий из списка продуктов
```

```python
train.ingredients=train.ingredients.apply(lambda x: ' '.join(x))
```

```python
train.ingredients 
```

```python
cv = TfidfVectorizer(max_df=0.99, min_df=0.001, ngram_range=(1,2))
features = cv.fit_transform(train.ingredients)
```

```python
len(cv.get_feature_names())
```

```python
features.toarray()
```

```python
# cv.get_feature_names()
```

```python
X = features
# y = train['F'].values
y = train.cuisine.values
y
```

```python
lr = LogisticRegression(multi_class='ovr')
lr.fit(X, y)
```

```python
y_pred = lr.predict(X)
y_pred
```

```python
is_action = lr.predict_proba(X)[:, 1]
is_action
```

```python
print('acc = ', metrics.accuracy_score(y, y_pred))

# counting the total true positives, false negatives and false positives
print('f1micro = ', metrics.f1_score(y, y_pred, average='micro')) 
#  для каждого ярлыка и найдите их невзвешенное среднее (При этом не учитывается дисбаланс )
print('f1macro = ', metrics.f1_score(y, y_pred, average='macro'))
# для каждого ярлыка средневзвешенные значения 
print('f1w = ', metrics.f1_score(y, y_pred, average='weighted'))
```
