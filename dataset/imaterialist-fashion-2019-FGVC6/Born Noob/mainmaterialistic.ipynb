```python
import cv2
import os
```

```python
def function_to_print_attributes_by_class_id(dataframe,train_dir,class_id,name,num_examples=2):
#     byClass_Dataframe = dataframe[dataframe["ClassId"]==class_id]
#     to_plt = byClass_Dataframe.head(randint(1,10))["ImageId"].values[0]
#     img = plt.imread(os.path.join(train_dir,to_plt))
#     plt.imshow(img)
    test_dataframe = dataframe["ClassId"].values
    for i in range(num_examples):
        random_num = choice([ i for (i,data) in enumerate(test_dataframe) if class_id in data])
        to_plt = dataframe[random_num:random_num+1]["ImageId"].values[0]
        img = cv2.imread(os.path.join(train_dir,to_plt))
        cv2.imwrite("./"+name +str(i)+".jpg",img)
        plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
attrib = data["attributes"]
for i,key in enumerate(attrib):
    if key["id"]==i:
        string = key["name"]
        function_to_print_attributes_by_class_id(df,train_dir,str(i),string,num_examples=2)
#         print(i,string)
```

```python
attributes = {}
for x in data["attributes"]:
    if x["supercategory"] not in attributes:
        attributes[x["supercategory"]] = []
        attributes[x["supercategory"]].append(x["name"])
    else:
        attributes[x["supercategory"]].append(x["name"])
```

```python
lists = []
names = []
for i in data["attributes"]:
    ids = i['id']
    lists.append(count(df,str(ids)))
    
    name = i["name"]
    names.append(name)
```
