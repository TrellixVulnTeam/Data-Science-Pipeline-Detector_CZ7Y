```python
data_india.groupby('days').sum()
data_india.drop(data_india.index[[0,1]] , inplace = True)
data_india.head()
```

```python
india_list = data_india["ConfirmedCases"].to_list()
```

```python
print(len(india_list))
print(india_list)
```

```python
india_list = [i for i in india_list if i>=28]
```

**To keep it simple using normal equation method.
and quadratic model to fit data to forcast.**

```python
print(len(india_list))
```

```python
india_arr = np.array(india_list)
y_train = (india_arr).T
print(y_train)
```

```python
a=[]
b=[]
for i in range(1,29):
    a.append(i**2)
    b.append(1)
arr = np.array(a)
brr = np.array(b)

x = np.vstack((brr , arr))
x_train = x.T


    
```

```python
print(len(x_train))
```

**Calculation of paramerter matrix**

```python
theta = np.linalg.inv((x).dot(x_train))
theta = theta.dot(x)
theta = theta.dot(y_train)
```

```python
print(theta)
```

```python
hyp = theta.T.dot(x)
```

```python
plt.plot(hyp)
plt.plot(india_list , "ro")
```
