```python
from fbprophet import Prophet
```

```python
df = pd.read_csv("/kaggle/input/covid19-global-forecasting-week-2/train.csv")
```

```python
import matplotlib.pyplot as plt
```

```python
def get_data_for_country(df, country_name, start_date=None):
    data = df[df['Country_Region'] == country_name]
    dates = sorted(list(set(data["Date"].to_numpy())))
    y = [ np.sum(data[data['Date'] == d]["ConfirmedCases"].to_numpy()) for d in dates]
    if start_date:
        ix = dates.index(start_date)
        dates = dates[ix:]
        y = y[ix:]
    dct = {"ds":dates, "y":y}
    return pd.DataFrame(dct)
```

```python
us_data = get_data_for_country(df, "US", start_date="2020-03-16")
```

```python
us_data
```

```python
def get_prophet_model(data):
    m = Prophet()
    m.fit(data)
    return m
```

```python
us_model = get_prophet_model(us_data)
```

```python
def get_future_predictions(m, period):
    future = m.make_future_dataframe(periods=period)
    forecast = m.predict(future)
    return forecast
```

```python
fut_us_preds = get_future_predictions(us_model, 8)
```

```python
us_model.plot(fut_us_preds)
```
