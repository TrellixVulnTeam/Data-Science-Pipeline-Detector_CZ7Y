TIMESERIES_MODEL

```python
data.columns=['ds','y']
m = Prophet(interval_width=0.95)
m.fit(data)
future = m.make_future_dataframe(periods=28)
forecast = m.predict(future)
forecast[['ds','yhat','yhat_lower','yhat_upper']].tail()
```

```python
d=m.plot(forecast)
```

```python
d1=m.plot_components(forecast)
```
