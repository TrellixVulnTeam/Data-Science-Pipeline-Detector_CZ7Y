```python
import os
import numpy as np 
import pandas as pd 
import seaborn as sns
import plotly.express as px
from plotly.offline import init_notebook_mode
import plotly.graph_objects as go
init_notebook_mode(connected=True)
pd.set_option('display.max_columns', 5000)
import warnings
warnings.filterwarnings("ignore")
#os.mkdir('/kaggle/working/individual_charts/')
import matplotlib.pyplot as plt
# Load the data

def load_csv(base_dir,file_name):
    """Loads a CSV file into a Pandas DataFrame"""
    file_path = os.path.join(base_dir,file_name)
    df = pd.read_csv(file_path,low_memory=False,encoding='ISO-8859-1')
    return df    

base_dir_2017 = '/kaggle/input/kaggle-survey-2017/'
file_name_2017 = 'multipleChoiceResponses.csv'
survey_df_2017 = load_csv(base_dir_2017,file_name_2017)
responses_df_2017 = survey_df_2017[1:]
survey_df_2017.to_csv('2017_kaggle_ds_and_ml_survey_responses_only.csv',index=False)

base_dir_2018 = '/kaggle/input/kaggle-survey-2018/'
file_name_2018 = 'multipleChoiceResponses.csv'
survey_df_2018 = load_csv(base_dir_2018,file_name_2018)
responses_df_2018 = survey_df_2018[1:]
survey_df_2018.to_csv('2018_kaggle_ds_and_ml_survey_responses_only.csv',index=False)

base_dir_2019 = '/kaggle/input/kaggle-survey-2019/'
file_name_2019 = 'multiple_choice_responses.csv'
survey_df_2019 = load_csv(base_dir_2019,file_name_2019)
responses_df_2019 = survey_df_2019[1:]
survey_df_2019.to_csv('2019_kaggle_ds_and_ml_survey_responses_only.csv',index=False)

base_dir_2020 = '/kaggle/input/kaggle-survey-2020'
file_name_2020 = 'kaggle_survey_2020_responses.csv'
survey_df_2020 = load_csv(base_dir_2020,file_name_2020)
responses_df_2020 = survey_df_2020[1:]
survey_df_2020.to_csv('2020_kaggle_ds_and_ml_survey_responses_only.csv',index=False)

base_dir_2021 = '/kaggle/input/kaggle-survey-2021/'
file_name_2021 = 'kaggle_survey_2021_responses.csv'
survey_df_2021 = load_csv(base_dir_2021,file_name_2021)
responses_df_2021 = survey_df_2021[1:]
survey_df_2021.to_csv('2021_kaggle_ds_and_ml_survey_responses_only.csv',index=False)

print('Total Number of Responses 2017: ',responses_df_2017.shape[0])
print('Total Number of Responses 2018: ',responses_df_2018.shape[0])
print('Total Number of Responses 2019: ',responses_df_2019.shape[0])
print('Total Number of Responses 2020: ',responses_df_2020.shape[0])
print('Total Number of Responses 2021: ',responses_df_2021.shape[0])
```

```python
#Kagglers demographics
r_dgc_2021 = responses_df_2021.iloc[:,0:7]
r_dgc_2020 = responses_df_2020.iloc[:,0:7]
r_dgc_2021['Finish_time'] = r_dgc_2021['Time from Start to Finish (seconds)'].astype('int')
r_dgc_2020['Finish_time'] = r_dgc_2020['Time from Start to Finish (seconds)'].astype('int')
```

```python
dg_c_2021.columns = ['Age','Gender','Country','Education','Employment','Experience','Numbers','Year']
dg_c_2020.columns = ['Age','Gender','Country','Education','Employment','Experience','Numbers','Year']
```

```python
#Kagglers demographics
r_dg2019 = responses_df_2019.iloc[:,0:12]
r_dg2019['Finish_time'] = r_dg2019['Time from Start to Finish (seconds)'].astype('int')
r_dg2019.drop(['Q2_OTHER_TEXT','Q5_OTHER_TEXT','Q9_Part_1','Time from Start to Finish (seconds)'],inplace=True,axis=1)
r_dg2019.drop(['Q6','Q8'],inplace=True,axis=1)
r_dg2019['Year'] = 2019
r_dg2019.columns = ['Age','Gender','Country','Education','Employment','Experience','Numbers','Year']
r_dg2019.head()
```

```python
#Kagglers demographics
r_dg2018 = responses_df_2018.iloc[:,0:12]
r_dg2018['Finish_time'] = r_dg2018['Time from Start to Finish (seconds)'].astype('int')
r_dg2018.drop(['Q7_OTHER_TEXT','Q6_OTHER_TEXT','Q1_OTHER_TEXT','Time from Start to Finish (seconds)'],axis=1,inplace=True)
r_dg2018.drop(['Q5','Q7'],axis=1,inplace=True)
r_dg2018['Year'] = 2018
r_dg2018.columns = ['Gender','Age','Country','Education','Employment','Experience','Numbers','Year']
r_dg2018 = r_dg2018[['Age','Gender','Country','Education','Employment','Experience','Numbers','Year']]
r_dg2018.head()
```

```python
#Kagglers demographics
r_dg2017 = responses_df_2017.iloc[:,0:12]
r_dg2017.drop(['CareerSwitcher','CodeWriter','LearningDataScience','StudentStatus','MLToolNextYearSelect',
              'CurrentEmployerType','TitleFit'],axis=1,inplace=True)
r_dg2017['Year'] = 2017
r_dg2017['Numbers'] = 1
r_dg2017.drop('EmploymentStatus',inplace=True,axis=1)
r_dg2017.columns = ['Gender','Country','Age','Employment','Year','Numbers']
r_dg2017 =r_dg2017[['Age','Gender','Country','Employment','Numbers','Year']]
r_dg2017.head()
```

```python
main_Dg = pd.DataFrame()
main_Dg = main_Dg.append(dg_c_2020)
main_Dg = main_Dg.append(dg_c_2021)
main_Dg = main_Dg.append(r_dg2019)
main_Dg = main_Dg.append(r_dg2018)
main_Dg = main_Dg.append(r_dg2017)
print(main_Dg.shape)
main_Dg.head()
```

```python
main_Dg.reset_index(inplace=True)
main_Dg.drop('index',inplace=True,axis=1)
```

```python
main_Dg.loc[main_Dg.Age.isna(),'Age'] = 'not_disclosed'
main_Dg.loc[(main_Dg.Age == 'NaN'),'Age'] = 'not_disclosed' 
main_Dg.loc[(main_Dg.Age == 'NA'),'Age'] = 'not_disclosed'
main_Dg['Age'] = main_Dg.Age.astype('str')
main_Dg.loc[main_Dg.Country.isnull(),'Country'] = 'not-disclosed'
main_Dg.loc[main_Dg.Education.isnull(),'Education'] = 'not-disclosed'
main_Dg.loc[main_Dg.Employment.isnull(),'Employment'] = 'not-disclosed'
main_Dg.loc[main_Dg.Experience.isnull(),'Experience'] = 'not-disclosed'
main_Dg.loc[(main_Dg.Experience == '3-4'),'Experience'] = '3-5'
main_Dg.loc[main_Dg.Numbers.isnull(),'Numbers'] = 1
#main_Dg.drop(['Finish_time'],inplace=True,axis=1)
main_Dg.loc[main_Dg.Gender.isna(),'Gender'] = 'Prefer not to say'
main_Dg.loc[main_Dg.Gender.str.contains('Non-binary'),'Gender'] = 'Nonbinary'
main_Dg.loc[main_Dg.Age.str.contains('-'),'Age'] = main_Dg.loc[main_Dg.Age.str.contains('-'),'Age'].apply(lambda x: x.split('-')[0])
main_Dg.loc[(main_Dg.Gender == 'Male'),'Gender'] = 'Man'
main_Dg.loc[(main_Dg.Gender == 'Female'),'Gender'] = 'Woman'
main_Dg.head()
```

### What change did the years 2019 to 2021 show in the age demographics?

```python
young_kagglers = counting_age.loc[(counting_age.Age == '18')| 
                 (counting_age.Age == '22')|
                 (counting_age.Age == '25')|
                 (counting_age.Age == '30')]

kaggler_2019 = young_kagglers.loc[young_kagglers.Year == 2019,'Numbers']
kaggler_2020 = young_kagglers.loc[young_kagglers.Year == 2020,'Numbers']
kaggler_2021 = young_kagglers.loc[young_kagglers.Year == 2021,'Numbers']

per_inc = (kaggler_2021.values - kaggler_2020.values)*100/kaggler_2020.values
per_red = (kaggler_2019.values - kaggler_2020.values)*100/kaggler_2019.values

change_show = make_subplots(rows=1,cols=2,shared_yaxes=True,horizontal_spacing=0.02)
change_show.add_trace(go.Bar(x=young_kagglers.Age.unique(), y = per_red,
                             name='Covid_% Reduction',text= per_red.round(1), textposition='auto'),row=1,col=1)
change_show.add_trace(go.Bar(x=young_kagglers.Age.unique(), y = per_inc,name='Post % Increase',
                             text= per_inc.round(1), textposition='auto'),row=1,col=2)
change_show.update_layout(height=500,width=800,xaxis = {'categoryorder':'total ascending'})
change_show.show()
```

2020 the age group that reduced by 62% was the age group between 18 to 20, and the same age group members returned with overwhelming 39% in 2021. Kaggling must have taken a back seat, as their studies moved to online, and the uncertainty in their lives increased.
