Some useful function I will use through this notebook.

```python
def answer(data, key1, key2='Count'):
    """
    This Function takes and counts values
    from all columns in given dataframe, than
    returns a new dataframe.
     """
    my_dict = {key1: list(),
               key2: list()}
    for q in data.columns:
        answer = data[q].value_counts()
        my_dict[key1].append(answer.index[0])
        my_dict[key2].append(answer.values[0])
    answer_df = pd.DataFrame(my_dict)
    return answer_df
```

```python
def show_barplot_v(data, col):
    """
    Function plots vertical barplot.
    
    """
    plt.figure(figsize=(12,7), dpi=130)
    sns.barplot(x=data.columns[1], y=data.columns[0], data=data.sort_values(by=data.columns[1], ascending=False), 
                palette='mako_r')
    plt.title('{}?'.format(df.iloc[:,col][0].split('?')[0]))
    
    
def show_barplot_h(data, col):
    """
    Function plots horizontal barplot
    """
    plt.figure(figsize=(12,5), dpi=130)
    sns.barplot(x=data.columns[0], y=data.columns[1], data=data, palette='mako_r')
    plt.title('{}?'.format(df.iloc[:,col][0].split('?')[0]))
    plt.xticks(rotation=45, horizontalalignment="right")
```
