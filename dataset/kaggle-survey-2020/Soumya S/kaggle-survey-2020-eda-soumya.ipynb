```python
k_data=k_data.iloc[1:, :]
```

First of all, lets see the gender distribution according to age group.

```python
plt.figure(figsize=(18,7))
sns.countplot(x=k_data.iloc[:,1],order=['18-21','22-24','25-29','30-34','35-39','40-44','45-49','50-54','55-59','60-69','70+'],hue=k_data.iloc[:,2])
```

Looks like Men are ahead in every age group. A shout out to all the Women to buck up!

Also, the 20s age group is more fasinated than the other groups with this field.

Lets see the qualification of the data professionals

```python
plt.figure(figsize=(20,7))
sns.countplot(y=k_data.iloc[:,4],order=['No formal education past high school','Some college/university study without earning a bachelor’s degree','Bachelor’s degree','Master’s degree','Doctoral degree','Professional degree','I prefer not to answer'])
```

So this graph shows that the data professionals have either a Bachelor's or Master's degree

This below graph shows the current title they are holding

```python
plt.figure(figsize=(20,7))
sns.countplot(y=k_data.iloc[:,5],)
plt.xlabel('')
```

From above 3 graphs I understand that most Data Science professionals are in there 20s and a student of either a Bachelor's or Master's degree.

Now let us see which country that has more number of Data professionals...

```python
plt.figure(figsize=(20,7))
k_data.iloc[:,3].value_counts().head(25).plot(kind="pie")
plt.show()
```

Pheww! India so far ahead of other countries when it comes to Data science. 

Lets see which are the Programming Languages are most used.

```python
fig,ax=plt.subplots(1,13,sharey=True,figsize=(20,10))
sns.countplot(x=k_data.iloc[:,7],ax=ax[0],color='red')
sns.countplot(x=k_data.iloc[:,8],ax=ax[1])
sns.countplot(x=k_data.iloc[:,9],ax=ax[2],color='yellow')
sns.countplot(x=k_data.iloc[:,10],ax=ax[3],color='green')
sns.countplot(x=k_data.iloc[:,11],ax=ax[4],color='black')
sns.countplot(x=k_data.iloc[:,12],ax=ax[5],color='orange')
sns.countplot(x=k_data.iloc[:,13],ax=ax[6])
sns.countplot(x=k_data.iloc[:,14],ax=ax[7],color='pink')
sns.countplot(x=k_data.iloc[:,15],ax=ax[8],color='violet')
sns.countplot(x=k_data.iloc[:,16],ax=ax[9],color='cyan')
sns.countplot(x=k_data.iloc[:,17],ax=ax[10],color='yellow')
sns.countplot(x=k_data.iloc[:,18],ax=ax[11],color='green')
sns.countplot(x=k_data.iloc[:,19],ax=ax[12])
for a in ax.flat:
    a.set(xlabel='', ylabel='',ylim=(0,16000))
```

Python dominates...

However, the below graph shows that there are some Python users that don't recommend Python to aspiring data scientist!!!

```python
plt.figure(figsize=(20,7))
k_data[k_data['Q7_Part_1']=='Python' ]['Q8'].value_counts().plot(kind="bar",color='violet')
plt.show()
```

That's approximately 19%
