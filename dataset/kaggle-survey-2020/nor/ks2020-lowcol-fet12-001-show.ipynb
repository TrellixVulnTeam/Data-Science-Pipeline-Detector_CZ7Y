## PCA

```python
wgt = np.r_[wgt_prod, wgt_user]
from sklearn import decomposition
pca = decomposition.PCA(n_components=12)
pca.fit(wgt)
```

```python
f = pca.transform(wgt)
f_prod = f[:wgt_prod.shape[0],]
f_user = f[wgt_prod.shape[0]:,]
```

```python
sns.pairplot(pd.DataFrame(f[:,:5]), markers=['.'], height=2.2)
save_img()
```
