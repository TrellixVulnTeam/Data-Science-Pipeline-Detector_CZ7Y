```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
data="../input/kaggle-survey-2020/kaggle_survey_2020_responses.csv"
dataset=pd.read_csv(data)
datasets=dataset.drop(0)
```

**Age/Gender Distribution**

```python
agecount=datasets["Q1"].value_counts()
agecountplot=agecount.plot(kind="bar",title="Count of age",xlabel="Agedistribution")
agecountplot.grid()

```

You can see that more than half of the total is under 30.

Look out that mostly from the age of 22 to 29 were mostly involved in datascience and machine learning concept.

18 to 21 age groups people were showing initial boost to the datascience and machine learning concept.

The interest towards this datascience and machine learning concept is gradually decreasing, from 34 to 50 age groups.

In the age group of 40's mostly they are very less interactive

18 to 30 is the best age group that lot of peoples are more interactive.

The e-commerce website,sales team,higher education admission team.They can target 18-30 age group people.Because the count of that group is huge.



**Gender distribution**


```python
gender=datasets[(datasets.Q2=="Man")| (datasets.Q2=="Woman")]
genders=gender[["Q1","Q2"]].value_counts()
genders.plot.bar("Q1","Q2",title="Agecount",xlabel="Gender and age distribution")



```

Obviously, I felt that there was a lot of influx of women, but I can see that there are still few.

The accessibility of a women is quite low in all age group.

Target 18 to 30 age group men is the best marketing strategy.One who is eagerly waiting to sell their products

From 40 to 60 age group woman is highly not recommended

### **Experience + Education**
