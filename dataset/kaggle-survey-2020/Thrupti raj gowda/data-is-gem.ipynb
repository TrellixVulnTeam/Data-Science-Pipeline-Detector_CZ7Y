```python

def proglang(df,n,max_row,label_name):
   
    lang = []
    count = []
    pl = data.loc[:,[f'Q{n}_Part_1']]
    for i in range(2,max_row+1):
        pl[f'Q{n}_Part_{i}'] = data.loc[:,[f'Q{n}_Part_{i}']]
    pl[f'Q{n}_OTHER'] = data.loc[:,[f'Q{n}_OTHER']]
        
    
    for j in range(1,max_row+1):
        for que in pl[f'Q{n}_Part_{j}'].dropna().unique():
            count.append(int(pl[f'Q{n}_Part_{j}'].shape - pl[f'Q{n}_Part_{j}'].isnull().sum()))
            lang.append(que)
    for q in pl[f'Q{n}_OTHER'].dropna().unique():
        count.append(int(pl[f'Q{n}_OTHER'].shape - pl[f'Q{n}_OTHER'].isnull().sum()))
        lang.append(q)
    d = {'count': count, 'lang': lang}
    plot_name = pd.DataFrame(data=d)
    plot_name['lang'] = plot_name['lang'].astype(str)
    plt.figure(figsize=(30,30))
    sns.set_style(style='darkgrid')
    total= float(len(plot_name['count']))
    ax = plot_name.plot(x='lang',y='count',kind='bar')
    plt.xticks(rotation=90)
    plt.xlabel(label_name)

    for p in ax.patches:
        height = p.get_height()
        ax.text(p.get_x()+p.get_width()/2.,height+450
               ,'{:4.0f}'.format(height,ha='center'))
```

```python
def proglangsub(alphabet,df,n,max_row,label_name):
   
    lan = []
    cnt = []
    pl = data.loc[:,[f'Q{n}_{alphabet}_Part_1']]
    for i in range(2,max_row+1):
        pl[f'Q{n}_{alphabet}_Part_{i}'] = data.loc[:,[f'Q{n}_{alphabet}_Part_{i}']]
    pl[f'Q{n}_{alphabet}_OTHER'] = data.loc[:,[f'Q{n}_{alphabet}_OTHER']]
        
    
    for j in range(1,max_row+1):
        for que in pl[f'Q{n}_{alphabet}_Part_{j}'].dropna().unique():
            cnt.append(int(pl[f'Q{n}_{alphabet}_Part_{j}'].shape - pl[f'Q{n}_{alphabet}_Part_{j}'].isnull().sum()))
            lan.append(que)
    for q in pl[f'Q{n}_{alphabet}_OTHER'].dropna().unique():
        cnt.append(int(pl[f'Q{n}_{alphabet}_OTHER'].shape - pl[f'Q{n}_{alphabet}_OTHER'].isnull().sum()))
        lan.append(q)
    d = {'count': cnt, 'lang': lan}
    plot_name = pd.DataFrame(data=d)
    plot_name['lang'] = plot_name['lang'].astype(str)
    plt.figure(figsize=(30,30))
    sns.set_style(style='darkgrid')
    total= float(len(plot_name['count']))
    ax = plot_name.plot(x='lang',y='count',kind='bar')
    plt.xticks(rotation=90)
    plt.xlabel(label_name)

    for p in ax.patches:
        height = p.get_height()
        ax.text(p.get_x()+p.get_width()/2.,height+450
               ,'{:4.0f}'.format(height,ha='center'))
```
