```python
#Importing libraries
import numpy as np
import pandas as pd
from plotly.offline import init_notebook_mode, iplot, plot
import plotly as py
init_notebook_mode(connected=True)
import plotly.graph_objects as go
import plotly.io as pio
from plotly.subplots import make_subplots
pio.templates.default = "plotly"
import plotly.express as px
from IPython.display import SVG
import pydot
colors = ['rgba(38, 24, 74, 0.8)', 'rgba(71, 58, 131, 0.8)']
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
df = pd.read_csv('/kaggle/input/kaggle-survey-2020/kaggle_survey_2020_responses.csv')
```

```python
df = df.iloc[1:, 1:]
df= df.fillna('')
```

```python
#list of programming languages preferred
df['programming_lang'] = df['Q7_Part_1']+'-'+df['Q7_Part_2']+'-'+df['Q7_Part_3']+'-'+df['Q7_Part_4']+'-'+df['Q7_Part_5']+'-'+df['Q7_Part_6']+'-'+df['Q7_Part_7']+'-'+df['Q7_Part_8']+'-'+df['Q7_Part_9']+'-'+df['Q7_Part_10']+'-'+df['Q7_Part_11']+'-'+df['Q7_Part_12']+'-'+df['Q7_OTHER']
df['programming_lang']=  df['programming_lang'].map(lambda x: x.split('-'))
for x in range(1, len(df['programming_lang'])+1):
    l= ''
    for elem in df['programming_lang'][x]:
        if elem.strip():
            l+=elem+','
    df['programming_lang'][x] = l[:-1]  
```

```python
#dropping unnecessary columns
df.drop(df.columns[6:19],axis=1,inplace=True)
df.drop(df.columns[7:51],axis=1,inplace=True)
df.drop(df.columns[8:51],axis=1,inplace=True)
df.drop(df.columns[15:17],axis=1,inplace=True)
df.drop(df.columns[16:137],axis=1,inplace=True)
df.drop(df.columns[17:131],axis=1,inplace=True)
```

```python
#dividing dataset into two age groups: 18-24 and 25+
df['Q1'] = df['Q1'].map(lambda x: '18-24' if x in ['18-21','22-24'] else '25+')
#Categorizing all other degrees other than bachelor's and master's as Others
df['Q4'] = df['Q4'].map(lambda x: 'Others' if x not in ['Bachelorâ€™s degree','Masterâ€™s degree'] else x)
#Creating a columns for employement status of each person
df['Employed'] = df['Q5'].map(lambda x: 'No' if x == 'Currently not employed' else 'Yes')
#Renaming value
df['Q6'] = df['Q6'].map(lambda x: '0 years' if x in ['I have never written code',''] else x)
#dividing dataset into two earning groups: <15000 and 15000+
df['Q24'] = df['Q24'].map(lambda x: '<15000' if x in ['$0-999', '1,000-1,999','2,000-2,999','3,000-3,999',
                                                      '4,000-4,999', '5,000-7,499','7,500-9,999','10,000-14,999',''] else '15000+')
```

```python
temp = pd.DataFrame(df.Q1.value_counts()*100.00/df.shape[0])

fig = go.Figure()

for i in range(len(temp)):
    fig.add_trace(go.Bar(
        y=[''],
        x=[temp.Q1[i]],
        name= temp.index[i],text= temp.Q1.index[i], textposition='auto',
        hovertemplate= "Percentage: %{x}%<extra></extra>",
        marker=dict(color=colors[i],line=dict(color='rgba(0, 0, 0)', width=2)),
        orientation='h' ))

fig.update_layout(barmode='stack', title = '<b>AGE DISTRIBUTION OF SAMPLE</b>', xaxis = dict(showgrid=False,title = 'Count'))
fig.show()
```

ðŸŸ¥ Chances of employement increases for people with Master's degree.

```python
temp = pd.DataFrame(df.Q4.value_counts())
yes = []
no=[]
for i in range (len(temp)):
    yes.append(list(df[df.Q4 ==temp.index[i]].Employed.value_counts()*100.00/df[df.Q4 ==temp.index[i]].shape[0])[0])
    no.append(list(df[df.Q4 ==temp.index[i]].Employed.value_counts()*100.00/df[df.Q4 ==temp.index[i]].shape[0])[1])
    
fig = go.Figure()

fig.add_trace(go.Bar(
    y= temp.index.tolist(),
    x= yes,
    name= 'Unemployed',text= 'Employed', textposition='auto',
    hovertemplate= "Percentage: %{x}%<extra></extra>",
    marker=dict(color=colors[0],line=dict(color='rgba(0, 0, 0)', width=2)),
    orientation='h' ))

fig.add_trace(go.Bar(
    y= temp.index.tolist(),
    x= no,
    name= 'Employed',text= 'Unemployed', textposition='auto',
    hovertemplate= "Percentage: %{x}%<extra></extra>",
    marker=dict(color=colors[1],line=dict(color='rgba(0, 0, 0)', width=2)),
    orientation='h' ))

fig.update_layout(barmode='stack', title = '<b>LEVEL OF FORMAL EDUCATION</b>', xaxis = dict(showgrid=False, title = '%'))
fig.show()
```

```python
dot_graph = pydot.Dot(graph_type='digraph')

root = pydot.Node('Survey Dataset')
root.set_shape('box')
root.set_style('filled')
dot_graph.add_node(root)

agenode = pydot.Node('18-24')
agenode.set_shape('record')
dot_graph.add_node(agenode)
agenode.set_label('<{{18-24|<font color="green">36.2%</font>}|{{Student|55.2%}|{Data Scientist|7.6%}|{Software Engineer|7.2%}|...|{DBA/Engineer|<font color="red">0.19%</font>}}}>')
iedge = pydot.Edge(root,agenode)
iedge.set_label('\t\t\t\tAGE')
dot_graph.add_edge(iedge)

new = pydot.Node('DBA/Engineer')
new.set_shape('box')
new.set_style('filled')
dot_graph.add_node(new)
iedge = pydot.Edge(agenode,new)
iedge.set_style('dashed')
dot_graph.add_edge(iedge)

agenode = pydot.Node('25+')
agenode.set_shape('record')
dot_graph.add_node(agenode)
agenode.set_label('<{{25+|<font color="green">63.7%</font>}|{{Data Scientist|16.6%}|{Software Engineer|11.2%}|{Student|9.06%}|...|{DBA/Engineer|<font color="red">0.86%</font>}}}>')
iedge = pydot.Edge(root,agenode)
dot_graph.add_edge(iedge)
iedge = pydot.Edge(agenode,new)
iedge.set_style('dashed')
dot_graph.add_edge(iedge)

agenode = pydot.Node('Age')
agenode.set_shape('record')
dot_graph.add_node(agenode)
agenode.set_label('<{{18-24|11.2%}|{25+|88.8%}}>')
iedge = pydot.Edge(new,agenode)
iedge.set_label('Age')
dot_graph.add_edge(iedge)

gennode = pydot.Node('Gender')
gennode.set_shape('record')
dot_graph.add_node(gennode)
gennode.set_label('<{{Men|82.4%}|{Women|12.8%}|{Non-binary|1.6%}|{Others|3.2%}}>')
iedge = pydot.Edge(new,gennode)
iedge.set_label('Gender')
dot_graph.add_edge(iedge)

salnode = pydot.Node('Salary')
salnode.set_shape('record')
dot_graph.add_node(salnode)
salnode.set_label('<{{15000|48.8%}|{15000+|51.2%}}>')
iedge = pydot.Edge(new,salnode)
iedge.set_label('Earn\n(USD)')
dot_graph.add_edge(iedge)

degnode = pydot.Node('Education')
degnode.set_shape('record')
dot_graph.add_node(degnode)
degnode.set_label("<{{Bachelor's degree|32.8%}|{Master's degree|44%}|{Others|23.2%}}>")
iedge = pydot.Edge(new,degnode)
iedge.set_label('Highest\ndegree')
dot_graph.add_edge(iedge)

usenode = pydot.Node('Use')
usenode.set_shape('record')
dot_graph.add_node(usenode)
usenode.set_label("<{Python|SQL|Bash|R|Javascript}>")
iedge = pydot.Edge(new,usenode)
iedge.set_label('Use')
dot_graph.add_edge(iedge)

recnode = pydot.Node('Recommend')
recnode.set_shape('record')
dot_graph.add_node(recnode)
recnode.set_label("<{Python|SQL|C|Javascript|R|Julia|Java}>")
iedge = pydot.Edge(new,recnode)
iedge.set_label('Recommend\nlearning')
dot_graph.add_edge(iedge)

dot_graph.write_svg('age_profession.svg')
SVG('age_profession.svg')
```
