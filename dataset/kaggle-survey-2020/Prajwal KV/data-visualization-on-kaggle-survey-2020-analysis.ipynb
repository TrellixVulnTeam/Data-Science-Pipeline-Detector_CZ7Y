# **What is your age (# years)?****


```python
sns.set_context('notebook',font_scale=2.3)
sns.set_style('whitegrid')
fig, ax1 = plt.subplots(figsize=(15,5))
graph = sns.countplot(x='Q1',palette='Paired',saturation=9.75,data=df,order=df['Q1'].value_counts().index)
graph.set_xticklabels(graph.get_xticklabels(),rotation=90)
for p in graph.patches:
    graph.annotate(format(p.get_height(), '.1f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   size=18,
                   xytext = (0,-7), 
                   textcoords = 'offset points')
plt.xlabel("Age Category of people who took Kaggle 2020 Surveys", size=25)
plt.ylabel("Count based on their Age category", size=25)
plt.title("Age wise count analysis of  Kaggle users")
plt.show()
```

## From the above Bar plot,we can find that most of the Kaggle users were between (25-29)Age    Groups Followed by (22-24),(18-21) and so forth.
