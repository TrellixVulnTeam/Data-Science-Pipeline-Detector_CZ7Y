```python
Q5Order = ['Business Analyst',
           'Data Analyst',
           'Data Engineer',
           'Data Scientist',
           'DBA/Database Engineer',
           'Machine Learning Engineer',
           'Product/Project Manager',
           'Research Scientist',
           'Software Engineer',
           'Statistician',
           'Student',
           'Currently not employed',
           'Other']

Q15Order = ['I do not use machine learning methods',
            'Under 1 year',
            '1-2 years',
            '2-3 years',
            '3-4 years',
            '4-5 years',
            '5-10 years',
            '10-20 years',
            '20 or more years']
```

```python
dataQ5Q15=data.groupby('Q5')['Q15'].value_counts().\
unstack(fill_value=0).loc[Q5Order,Q15Order]
dataQ5Q15man=data[data['Q2']=='Man'].groupby('Q5')['Q15'].\
value_counts().unstack(fill_value=0).loc[Q5Order,Q15Order]
dataQ5Q15woman=data[data['Q2']=='Woman'].groupby('Q5')['Q15'].\
value_counts().unstack(fill_value=0).loc[Q5Order,Q15Order]
```

```python
def drawPie(xp,yp,ratios,sizes,colors,ax):
    markers=[]
    before=0
    for color,ratio in zip(colors,ratios):
        after=2*np.pi*ratio+before
        x=[0]+np.cos(np.linspace(before,after,50)).tolist()
        y=[0]+np.sin(np.linspace(before,after,50)).tolist()
        before=after
        xy=np.column_stack([x,y])
        markers.append({'marker':xy,'s':np.abs(xy).max()**2*np.array(sizes),
                       'facecolor':color})
    for marker in markers:
        ax.scatter(xp,yp,**marker,alpha=0.7)
```

```python
fig=plt.figure(figsize=(20,23),dpi=200)
gs=fig.add_gridspec(5,5)

ax_scatter=fig.add_subplot(gs[1:4,0:4])
for Q5_idx in Q5Order[::-1]:
    for Q15_idx in Q15Order:
        man=dataQ5Q15man.loc[Q5_idx,Q15_idx]
        woman=dataQ5Q15woman.loc[Q5_idx,Q15_idx]
        total=dataQ5Q15.loc[Q5_idx,Q15_idx]
        drawPie([Q15_idx],[Q5_idx],[man/total,woman/total],[total*2.5],ax=ax_scatter,
               colors=['#41a1f0','#f5536e'])
```

```python
fig=plt.figure(figsize=(20,23),dpi=200)
gs=fig.add_gridspec(5,5)

ax_scatter=fig.add_subplot(gs[1:4,0:4])
for Q5_idx in Q5Order[::-1]:
    for Q15_idx in Q15Order:
        man=dataQ5Q15man.loc[Q5_idx,Q15_idx]
        woman=dataQ5Q15woman.loc[Q5_idx,Q15_idx]
        total=dataQ5Q15.loc[Q5_idx,Q15_idx]
        drawPie([Q15_idx],[Q5_idx],[man/(man+woman),woman/(man+woman)],
                     [total*2.5],['#41a1f0','#f5536e'],ax=ax_scatter)

ax_scatter.set_xticklabels(labels=Q15Order,rotation=90,
                          fontsize=15,fontweight='bold')
ax_scatter.set_yticklabels(labels=Q5Order[::-1],
                          fontsize=15,fontweight='bold')


ax_up=fig.add_subplot(gs[0,:4],sharex=ax_scatter)
woman_up=data[data['Q2']=='Woman']['Q15'].value_counts().loc[Q15Order]
man_up=data[data['Q2']=='Man']['Q15'].value_counts().loc[Q15Order]

ax_up.bar(x=woman_up.index,height=woman_up,width=0.5,color='#f5536e')
ax_up.bar(x=man_up.index,height=man_up,width=0.5,color='#41a1f0',bottom=woman_up)


ax_side=fig.add_subplot(gs[1:4,4],sharey=ax_scatter)
woman_side=data[data['Q2']=='Woman']['Q5'].value_counts().loc[Q5Order]
man_side=data[data['Q2']=='Man']['Q5'].value_counts().loc[Q5Order]

ax_side.barh(y=woman_side.index[::-1],width=woman_side[::-1],height=0.5,
             color='#f5536e')
ax_side.barh(y=man_side.index[::-1],width=man_side[::-1],height=0.5,
             color='#41a1f0',left=woman_side[::-1])

plt.setp(ax_side.get_yticklabels(), visible=False)
plt.setp(ax_up.get_xticklabels(),visible=False)

for s in ['top','right','bottom','left']:
    ax_scatter.spines[s].set_visible(False)
    ax_up.spines[s].set_visible(False)
    ax_side.spines[s].set_visible(False)

    
fig.text(x=0.15,y=0.9,s='Q2 & Q5 & Q15 pie chart with stacked bar',
        fontweight='bold',fontsize=35)
plt.show()
```

```python
fig,axs=plt.subplots(6,2,figsize=(20,24),dpi=200)

for i in range(11):
    axs[i//2][i%2].bar(x=man.loc[Q5Order[i]].index,height=man.loc[Q5Order[i]],
                      color='#eb3a34',edgecolor='black',label='Man')
    axs[i//2][i%2].bar(x=woman.loc[Q5Order[i]].index,height=-woman.loc[Q5Order[i]],
                      color='#34b1eb',edgecolor='black',label='Woman')
    axs[i//2][i%2].set_title(Q5Order[i],
                            fontsize=17,fontweight='bold')
    axs[i//2][i%2].set_xticklabels([])
    axs[i//2][i%2].set_yticks([])
    axs[i//2][i%2].legend()

axs[-1][0].set_xticklabels(labels=compOrder,rotation=90)
axs[-1][1].set_visible(False)
plt.show()
```

<h1 style="background-color:black;color:white"><center>Q4 & Q5 & Q24</center></h1>

<h3>jobs and degree and compensation<br>
data engineer and DBA engineer and product_manager</h3>

```python
master=data[data['Q4']=='Master’s degree'].groupby('Q5')['Q24']\
.value_counts().unstack(fill_value=0)[compOrder]
bachelor=data[data['Q4']=='Bachelor’s degree'].groupby('Q5')['Q24']\
.value_counts().unstack(fill_value=0)[compOrder]
doctoral=data[data['Q4']=='Doctoral degree'].groupby('Q5')['Q24']\
.value_counts().unstack(fill_value=0)[compOrder]
professional=data[data['Q4']=='Professional degree'].groupby('Q5')['Q24']\
.value_counts().unstack(fill_value=0)[compOrder]
college=data[data['Q4']=='Some college/university study without earning a bachelor’s degree'].\
groupby('Q5')['Q24'].value_counts().unstack(fill_value=0)
college['> $500,000']=0
college=college[compOrder]
noformal=data[data['Q4']=='No formal education past high school'].\
groupby('Q5')['Q24'].value_counts().unstack(fill_value=0)
noformal[['> $500,000', '250,000-299,999', '125,000-149,999', '200,000-249,999', '300,000-500,000']]=0
noformal=noformal[compOrder]
```

```python
fig,axs=plt.subplots(6,2,figsize=(20,24),dpi=200)

for i in range(11):
    one=master.loc[Q5Order[i]]
    two=bachelor.loc[Q5Order[i]]
    three=doctoral.loc[Q5Order[i]]
    four=professional.loc[Q5Order[i]]
    five=college.loc[Q5Order[i]]
    six=noformal.loc[Q5Order[i]]
    
    axs[i//2][i%2].bar(x=one.index,height=one,
                      color='#eb3a34',edgecolor='black',label='Master degree')
    axs[i//2][i%2].bar(x=two.index,height=two,
                      color='#34b1eb',edgecolor='black',
                       label="barchelor's degree",bottom=one)
    axs[i//2][i%2].bar(x=three.index,height=three,
                      color='#a15e5d',edgecolor='black',
                       label='doctoral',bottom=one+two)
    axs[i//2][i%2].bar(x=four.index,height=four,
                      color='#e69447',edgecolor='black',
                       label='professional',bottom=one+two+three)
    axs[i//2][i%2].bar(x=five.index,height=five,
                      color='#9cbf28',edgecolor='black',
                       label='college',bottom=one+two+three+four)
    axs[i//2][i%2].bar(x=six.index,height=six,
                      color='#c726a1',edgecolor='black',
                       label='no formal education',bottom=one+two+three+four+five)
    axs[i//2][i%2].set_title(Q5Order[i],
                            fontsize=17,fontweight='bold')
    axs[i//2][i%2].set_xticklabels([])
    axs[i//2][i%2].set_yticks([])
    axs[i//2][i%2].legend()

axs[-1][0].set_xticklabels(labels=compOrder,rotation=90)
axs[-1][1].set_visible(False)
plt.show()
```
