```python
def visualize_data_pie(df):   
    if isinstance(df, pd.DataFrame):
        df.drop(index=0, inplace=True)
        labels=[]
        sizes=[]
        for col in df.columns:
                labels.append(''.join(df[col].value_counts().index.to_list()))
                sizes.append(df[col].value_counts()[0])
    else:
        title=df.loc[0]
        df.drop(index=0, inplace=True)
        sizes=df.value_counts().values
        labels=list(df.value_counts().index)
    
    fig,ax=plt.subplots(figsize=(12,10))
    ax.pie(sizes, labels=labels, autopct='%1.1f%%',shadow=True)
    ax.axis('equal')
    plt.show()
```

```python
def visualize_bar(val):
    title=val.loc[0]
    val.drop(index=0, inplace=True)
    plt.figure(figsize=(15,12))
    sns.countplot(val)

    plt.xticks(
        rotation=45, 
        horizontalalignment='right',
        fontweight='light',
        fontsize='x-large'  
    )
    plt.show()
```

```python
def visualize_data(df_r, col1):
    
    title=df_r.loc[0]
    df_r.drop(index=0, inplace=True)

    labels=[]
    sizes=[]
    for col in df_r.columns:
            labels.append(''.join(df_r[col].value_counts().index.to_list()))
            sizes.append(df_r[col].value_counts()[0])


    df=pd.DataFrame()
    df[col1]=labels
    df['Counts']=list(map(int, sizes))
    plt.figure(figsize=(12,10))
    sns.barplot(x='Counts',y=col1,data=df)
    plt.xticks(
        rotation=45, 
        horizontalalignment='right',
        fontweight='light',
        fontsize='x-large'  
    )
    plt.show()
```

# Average time spent with respect to country

```python
time_cty=data[['Time from Start to Finish (seconds)','Q3']].copy()
time_cty.columns=['Time','Country']
time_cty.drop(index=0, inplace=True)
time_cty.Time=time_cty['Time'].astype('int64')

plt.figure(figsize=(15,12))
time_cty.groupby('Country')['Time'].mean().sort_values(ascending=True).plot(kind='barh', color=['k','m','y'])
plt.xlabel('Avegare time spent(secs)')
plt.show()
```

Average time spent by participants from korea is higher than the rest.

# Relationship between gender and education of the participants.

```python
edu_gen=data[['Q4','Q2']].copy()
edu_gen.columns=['Education','Gender']
edu_gen.drop(index=0, inplace=True)


plt.figure(figsize=(15,10))
sns.countplot(y='Education',data=edu_gen, hue='Gender')
plt.xticks(
        rotation=45, 
        horizontalalignment='right',
        fontweight='light',
        fontsize='x-large'  
    )
plt.legend()
plt.show()
```

There are more men with master's degree than women. And men outnumber women in all the categories of education.
