Main questions of these dataset, That help us alot to tell us which country is more insterested in Data sciece and percentage of age of that country
- Q1 : What is your age (# years) \n
- Q2 : What is your gender?
- Q3 : In which country do you currently reside?

**Question 1**

```python
# iloc 0 contain question so i start dataframe with 1
ax=df.iloc[1:]['Q1'].value_counts().plot(kind='bar',title='Age that are involve in Data Science')
ax.set_xlabel("Age Range")
ax.set_ylabel("Sum of occurance")
```

**Question 2**

```python
ax=df.iloc[1:]['Q2'].value_counts().plot(kind='bar',title='Gender that are involve in Data Science',figsize=[20, 10])
ax.set_xlabel("Gender")
ax.set_ylabel("Sum of occurance")
```

**Question 3**

```python
ax=df.iloc[1:]['Q3'].value_counts().plot(kind='bar',title='In which country do you currently reside',figsize=[20, 10])
ax.set_xlabel("Countries")
ax.set_ylabel("Number of Peoples")
```
