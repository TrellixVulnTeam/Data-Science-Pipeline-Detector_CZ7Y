## List of question that are asked in the survey

```python
for i in data.loc[0]:
    print(i)
```

```python
questions=data.loc[0]
```

```python
data=data.drop(0)
```

```python
for col in data.columns:
  if data[col].nunique() == 1:
    values=data[col].unique()
    if pd.isna(values[0]):
      col_name='_'.join([col.split('_')[0],data[col].unique()[1].lstrip().rstrip().replace(' ','-')])
    else:
      col_name='_'.join([col.split('_')[0],data[col].unique()[0].lstrip().rstrip().replace(' ','-')])
    data[col]=data[col].apply(lambda x : 0 if pd.isna(x) else 1 )
    data.rename(columns={col: col_name},inplace=True)
```

```python
for ind in questions.index:
  if '_' not in ind:
    # print(ind,":",questions[ind])
    print("\n",questions[ind])
    print(data[ind].value_counts())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
for ind in questions.index:
  if '_' not in ind:
    # print(ind,":",questions[ind])
    plt.style.use('seaborn-whitegrid')
    fig = plt.figure(figsize=(20,7))
    ax = plt.axes()
    plt.xlabel(questions[ind])
    ax.plot(data[ind].value_counts().index, data[ind].value_counts())
#     sns.barplot(x=data[ind].value_counts(), y=data[ind].value_counts().index, orient="v", data=data, estimator=lambda x: len(x) / len(data) * 100)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

```python
qSets=[]
for i in range(1,35):
    mstr='Q'+str(i)
#     print([x for x in data.columns if x.split('_')[0] == mstr])
    qSets.append([x for x in data.columns if x.split('_')[0] == mstr])
```

```python
for i in qSets:
    if len(i)==1:
        plt.figure(figsize=(16,6))
#         print(i[0])
        ln_grph = sns.barplot(x=i[0], y=i[0], orient="v", data=data, estimator=lambda x: len(x) / len(data) * 100)
        ln_grph.set(ylabel="Percent")
```

```python
for i in qSets:
    if len(i)>1:
        tmp=i+['Q1']
        print(data[tmp].sum())
```

```python
#Plot graph to get insight of defaulters percentage
plt.figure(figsize=(16,6))
ln_grph = sns.barplot(x="Q2", y="Q1", orient="v", data=data, estimator=lambda x: len(x) / len(data) * 100)
ln_grph.set(ylabel="Percent")
```

```python
for i in qSets:
    if len(i)==1:
        temp = data.groupby(['Q1',i[0]])[i[0]].count()
        tmp=temp / temp.groupby(level=0).sum()*100

        data1=pd.DataFrame()
        for l in data.Q1.unique():
            data1=data1.append(tmp[l])

        data1['Q1']=data.Q1.unique()
        data1.set_index('Q1',inplace=True)
        #print(data.index)
        plt.figure(figsize=(20,6))
        grph = data1.sort_index().plot.barh(figsize=(20,7),stacked=True)
        grph.set_xlabel("Percentage")
        grph.set_ylabel("Age")
```
