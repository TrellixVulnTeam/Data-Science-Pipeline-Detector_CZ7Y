### Consideration: Respondents with generater than 2 years of experience on machine learning methods as experienced ML professionals.

```python
response.Q15.value_counts()
```

```python
experience=['2-3 years', '3-4 years', '4-5 years','5-10 years','10-20 years','20 or more years']
```

```python
mlprofessionals=response[response["Q15"].isin(experience)]
```

```python
mlprofessionals.head()
```

```python
def cc(df,column_name):#used to count the answers to a unique question.
    count=0
    for column in df.columns:
        for line in re.findall(column_name,column):
            count=count+1
    return count
```

```python
def col_num(df,column_name):#to count  survey results.
    k=cc(df,column_name + ".+")
    data=[]
    count=[]
    per=[]
    for i in range(1,k):
        data_count=(df[column_name + "Part_"+ "{}".format(i)].value_counts())
        data.append(data_count.index[0])
        count.append(data_count.values[0])
    other=df[column_name+"OTHER"].value_counts()
    data.append(other.index[0])
    count.append(other.values[0])
    for i in count:
        k=(np.round_((i/sum(count))*100,2))
        per.append(k)
    return data,count,per
```

# Cloud Computing Platforms

```python
data,count,percentage=col_num(mlprofessionals,"Q26_A_")
d={}
for x,y in zip(data,percentage):
    d[x]=y
    
colors=[]
for i in data:
    if re.search("Amazon", i):
        colors.append("#FF9900")
    elif re.search("AWS", i):
        colors.append("#FF9900")
    elif re.search("Azure", i):
        colors.append("#007fff")
    elif re.search("Google", i):
        colors.append("#0F9D58")
    else:
        colors.append("lightgrey")
# ----------------------------------------------------------------------------------------------------

fig = plt.figure(figsize = (14, 8))
ax = fig.add_subplot()

# ----------------------------------------------------------------------------------------------------

ax.bar(d.keys(), d.values(), color=colors,edgecolor = "black")


for i, (k, v) in enumerate(d.items()):
    ax.text(k, 
            v +0.8,
            v, 
            color ="black", 
            fontsize = 12, 
            horizontalalignment = 'center', 
            verticalalignment = 'center'
           )

# ----------------------------------------------------------------------------------------------------

ax.tick_params(axis = 'x', labelrotation = 90, labelsize = 12)
ax.tick_params(axis = 'y', labelsize = 12)


ax.set_ylim(0, 35)


ax.set_xlabel("Cloud computing platforms",fontsize=16)
ax.set_ylabel("Percentage of respondents",fontsize=16)

ax.set_title("Cloud Computing Platforms that ML professionals use on regular basis", fontsize = 18)
```
