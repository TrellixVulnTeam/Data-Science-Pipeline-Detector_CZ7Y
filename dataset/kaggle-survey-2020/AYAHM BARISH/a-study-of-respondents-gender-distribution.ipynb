Let's talk in more detail about age and how it is distributed by gender
However, since the number of men is much greater than the number of women, we will study the distribution of ages in the sample of men and the sample of women separately

```python
data_man=df[df['Q2']=="Man"]
data_woman=df[df['Q2']=="Woman"]
n_man=data_man.shape[0]
n_woman=data_woman.shape[0]
rat=data_man.shape[0]/data_woman.shape[0]
rat
```

# The number of men is four times the number of women

Let's count the number of women and men under the age of 30 years

```python
n1=data_woman[(data_woman.Q1=="18-21") | (data_woman.Q1=="22-24") | (data_woman.Q1=="25-29") |(data_woman.Q1=="30-34")].shape[0]
n2=data_man[(data_man.Q1=="18-21") | (data_man.Q1=="22-24") | (data_man.Q1=="25-29") |(data_woman.Q1=="30-34")].shape[0]
```

```python
print("The percentage of women under 35 years old:",n1/n_woman)
print("The percentage of men under 35 years old:",n2/n_man)
```

We obtained a result contrary to the assumptions based on the number of respondents, as the percentage of women under 35 years of age was greater than the percentage of men.

```python
sub_data=df[(df['Q2']=='Man')|(df['Q2']=="Woman")]
pal = ["royalblue", "dodgerblue", "lightskyblue", "lightblue"]
ax= pd.crosstab(sub_data['Q2'], sub_data['Q1']).apply(lambda r: r/r.sum()*100, axis=1)
ax_1 = ax.plot.bar(figsize=(20,10),stacked=True, rot=0)
display(ax)

plt.legend(loc='upper center', bbox_to_anchor=(0.5, 0.95), title="Age")

plt.xlabel('Gender')
plt.ylabel('Percent Distribution')

for rec in ax_1.patches:
    height = rec.get_height()
    ax_1.text(rec.get_x() + rec.get_width() / 2, 
              rec.get_y() + height / 2,
              "{:.0f}%".format(height),
              ha='center', 
              va='bottom')

plt.show()
```

> We also noted that more than 77% of women are under the age of 30, which is a remarkable result that we will try to explain based on other indicators later.
