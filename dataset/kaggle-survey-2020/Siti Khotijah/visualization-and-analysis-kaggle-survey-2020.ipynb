# Barplots

showing the frequency of each category separated by label

```python
# showing the frequency of each category separated by label
plt.figure(figsize=[15,17])
fft=['Q1','Q2','Q3','Q4','Q5','Q6']
n=1
for f in fft:
    plt.subplot(4,2,n)
    sns.countplot(x=f, hue='Q1', edgecolor="black", alpha=0.7, data=kaggler)
    sns.despine()
    plt.title("Countplot of {}  by Q1".format(f))
    n=n+1
plt.tight_layout()
plt.show()


    
plt.figure(figsize=[15,4])
sns.countplot(x='Q1', hue='Q3',edgecolor="black", alpha=0.7, data=kaggler)
sns.despine()
plt.title("Countplot of Q1 by Q3")
plt.show()
```

# See Kaggler by gender

```python
def kaggler_2020(x):
    y = kaggler[["Q2","Q3","Q4","Q5"]][kaggler["Q2"] == x]
    y = y.sort_values(by="Q3",ascending=False)
    return y.head(500)
```

```python
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objs as go

kg = kaggler['Q2'].value_counts().reset_index()
kg.columns = [
    'Q2', 
    'percent'
]
kg['percent'] /= len(kaggler)

fig = px.pie(
    kg, 
    names='Q2', 
    values='percent', 
    title='See Kaggler by gender', 
    width=800,
    height=500 
)

fig.show()
```

```python
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objs as go

kg = kaggler['Q2'].value_counts().reset_index()
kg.columns = [
    'Q2', 
    'count'
]
kg['Q2'] = kg['Q2'].astype(str) + '-'
kg = kg.sort_values(['count']).tail(50)

fig = px.bar(
    kg, 
    x='count', 
    y='Q2', 
    orientation='h', 
    title='See Kaggler by gender', 
    width=1000,
    height=900 
)

fig.show()
```

```python
kaggler_q2 = kaggler['Q2'].value_counts()  
plt.figure(figsize=(5,3))
sns.barplot(kaggler_q2.index, kaggler_q2.values, alpha=0.8)
plt.ylabel('Number of Q2', fontsize=8)
plt.xlabel('What is your gender? - Selected Choice', fontsize=7)
plt.xticks(rotation=90)
plt.show();
```

# What type of computing platform do you use most often for your data science projects?

```python
# Start with one description:
text = kaggler.Q11[1]

# Create and generate a word cloud image:
wordcloud = WordCloud().generate(text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
# lower max_font_size, change the maximum number of word and lighten the background:
wordcloud = WordCloud(max_font_size=30, max_words=100, background_color="white").generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
