# **Importing Required Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**EDA**

```python
df=pd.read_csv('../input/kaggle-survey-2020/kaggle_survey_2020_responses.csv')
df.head()
```

# **Setting Questions as index**

```python
for i in range(0,len(df.columns)):
    df.rename(columns={df.columns[i]:df[df.columns[i]][0]},inplace=True)
```

```python
df=df.iloc[1:]
```

```python
df.head()
```

# **Cleaning the questions**

```python
df.columns=[df.columns[i].split('?')[0] for i in range(0,len(df.columns))]
df.columns=[df.columns[i].split(':')[0] for i in range(0,len(df.columns))]
ex=pd.DataFrame({'columns':df.columns})
```

```python
ex
```

```python
len(ex['columns'].unique())
ex=ex.iloc[1:]
```

# **47 Distinct Questions are present**

```python
questionair=ex['columns'].unique()
len(questionair)
```

# **Total No of Unique quesions**

```python
for i in range(0,len(questionair)):
    print(questionair[i])
```

# **For Better understanding of code breaking the code into parts is advisible**

```python
j=1
if type(df[questionair[j]])==pd.core.series.Series:
  sns.barplot(df[questionair[j]].value_counts().values,df[questionair[j]].value_counts().index)
  plt.title(questionair[j])
else:
  popularity=pd.DataFrame({'Programming_language':[df[questionair[j]].iloc[:,i].value_counts().index for i in range(0,len(df[questionair[j]].columns))],'values':[df[questionair[j]].iloc[:,i].value_counts().values for i in range(0,len(df[questionair[j]].columns))]})
  language=pd.DataFrame({'ide':[popularity['Programming_language'][i][0] for i in range(0,len(df[questionair[j]].columns))],'Users':[popularity['values'][i][0] for i in range(0,len(df[questionair[j]].columns))]})  
  language=language.nlargest(len(df[questionair[j]].columns),'Users')
  sns.barplot(language['Users'],language['ide'])
  plt.title(df[questionair[j]].columns[0])
  plt.xlabel('Users')
  plt.ylabel('Y-Axis')
```
