# Overview

This is the 4th instalment in the series of surveys conducted by Kaggle since 2017. In 2020, there were a total of 20,036 responses captured across 355* questions. In this years' survey, supplementary questions were also asked to gauge interest in learning/training of various Cloud related, Bigdata, BI technologies, and  Machine learning tools.

Note: There are 355 questions because questions with multiple choice answers were considered as separate questions. Otherwise, there were 39 primary questions and 9 supplementary questions.

# Objective

This analysis tries to uncover insights from Kaggle surveys conducted in the last 3 years along several parameters such as Demographics, Work experience, and related attributes, and Computing Hardware, Cloud and Bigdata related parameters . It also does a comparison/contrast between the top 2 responding countries - India and USA, which together contribute to `8088` (40%`) of the total responses in 2020.

# Conclusion

The trend analysis threw up some interesting insights. A few of the noteable ones are:

**Demograhic trends**

1. The YoY survey responses have declined at the rate of `8%` - this could be attributed to the pandemic.
2. The top 10 countries in 2020 includes a new entrant - Nigeria.
3. India has the highest number of 18-21 year old Kagglers, whereas for the rest of the world, it is 25-29 years and for the USA, it is 30-34 years.
4. While India produces a higher percentage of Bachelor's degree holders, the US and the rest of the world have a relatively higher percentage of Master's degree holders.
5. Overall participation by Males has dropped at the rate of `11%` annually, whereas, Female participation has dropped much slower at the rate of `4%` annually.
6. In India, PhD holders have been growing at the rate of `10%` annually; this is contrary to the global decline of PhD holders, which is happening at the rate of `16%` annually.
7. Most Youngsters (8-19 years) from India are either Students or Software Engineers. In contrast, most Youngsters from the rest of the world are either Students or Data Scientists.
8. There is a huge income divide between Indians and US. Among the Indian Kagglers, only `7%` earn `$60K` and above annually. In comparison, `80%` of Americans earn `$60K` and above annually.

**Coding, languages and company analysis**

1. Most Young coders ( <= 2 years) of experience are found in India. In contrast, most Middle level (3-10 years) of experience are found in the rest of the world and in the USA. Additionally, the 20+ year
category has seen a tremendous growth rate of `88%` annually.
2. When it comes to having Machine learning experience. Most of the respondents are Juniors ( < 2 years).
3. Python is the still the most popular language that is used regularly; this is followed by SQL and R in that order.
4. The most popular recommended language is also Python. However, the second most popular recommended language is R and not SQL.
5. Spending on Machine learning services has not picked up much in the last 2 years. Almost half of the respondents have spent less than `$100` in the last 5 years.
6. India and USA each have the highest number of Large sized (1000+) employees than the rest of the globe.
7. There appears to be a direct correlation between the number of Data Science folks and the Company size. i.e, Large sized (1000+) see more individuals involved in Data Science workloads.
8. The most common Machine learning activity is to "Analyze and understand data to influence product or business decisions" indicating that Machine learning (in the truest sense) adoption rates are very low.

**Computing hardware, Notebooks, IDE and Cloud and Bigdata**

1. Most respondents still work using either PC or a Laptop.
2. Google colab is the most popular Hosted NB in 2020 knocking out Kaggle, which was at #1 for the last 2 years.
3. Jupyter is still the most popular IDE among the respondents. PyCharm has been significant inroads in the last 3 years gaining 3 places to #2 in 2020.
4. AWS and GCP the 2 popular Cloud computing resources that respondents use.
5. MySQL is the still the database of choice.

Some other observations:

* Kaggle asked a question around the use of specialized hardware on a regular basis, which doesn't provide an option for CPUs. I think CPUs are still very relevant. While not shown here, 'None' was the second most 
common response to this question. A similar question was asked in 2019, which included CPUs as one of the options.
* A few questions saw `60%` or more respondents ignoring them such as - 1) Spending on Machine learning services. 2) Company size. 3) Data Science workloads; there are multiple choice questions to which saw a drop 
in responses.
* Questions with a dozen or so options will only confuse the respondents and it might also induce fatigue. Something to take note of.
