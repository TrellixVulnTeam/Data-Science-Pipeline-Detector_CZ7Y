```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import pandas_bokeh as pb
pb.output_notebook()
import seaborn as sns
```

```python
responses= pd.read_csv('../input/kaggle-survey-2020/kaggle_survey_2020_responses.csv')
columns = responses.iloc[0,:]
responses.columns = columns
responses = responses.drop(responses.index[0])
```

## AGE OF USERS 

```python
ages = responses['What is your age (# years)?'].value_counts()
ages = ages.sort_index()
ages.name = 'Years'
```

### Seaborn

```python
plt.figure(figsize=(12,6))
plt.title('Age distribution of kaggle users')
sns.barplot(x = ages.index, y=ages)
```

### Bokeh

```python
ages.plot_bokeh(kind='bar',
               ylabel='Number of people')
```

## CODING LANGUAGES

```python
prog_l = [c for c in responses.columns if 'What programming languages do you use on a regular basis? (Select all that apply) - Selected Choice' in c]
```

```python
lista=[]
indexs = []
for i in range(0,len(prog_l)):
    indexs =  np.append(indexs,prog_l[i].split(' - Selected Choice - ')[1])
    lista = np.append(lista,responses[prog_l[i]].value_counts())
```

```python
programming_l = pd.DataFrame(data = {'program':indexs, 'value':lista})
```

### seaborn

```python
plt.figure(figsize=(14,6))
sns.barplot(x = programming_l['program'], y = programming_l['value'])
```
