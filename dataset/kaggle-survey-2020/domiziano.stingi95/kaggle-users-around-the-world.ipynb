Plot of the kaggle users around the world using GeoPandas

```python
import numpy as np 
import pandas as pd 
import geopandas as gpd
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
def plot_users(method):
    if method != 'Geopandas' and method != 'plotly':
        print('Choose between \"Geopandas\" and \"plotly\"')
    responses = pd.read_csv('../input/kaggle-survey-2020/kaggle_survey_2020_responses.csv',low_memory=False)
    countries = responses['Q3'].value_counts()[:-1]
    countries = pd.DataFrame(countries)
    count = countries.index
    countries['name'] = count
    countries.index = range(0,55)
    world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
    world = pd.merge(world,countries,on=['name'],how='left') 

    #########################
    
    if method == 'Geopandas':
        fig,ax=plt.subplots(figsize=(20,20))
        world.plot(column=world['Q3'],ax=ax,legend=True,
                   legend_kwds={
                   'label': "intensity",
            'orientation': "horizontal"},
                           missing_kwds={
                    'color':'lightgrey',
                    'label':'Missing values',
                    'edgecolor':'black'
                  } 


              )
        ax.set_title("PERCENTAGE WORLD DISTRIBUTION OF KAGGLE USERS", fontsize=25)
    
    if method == 'plotly':
        fig = px.choropleth(world, locations=world['name'],
                    color_continuous_scale="Viridis",
                    locationmode='country names',
                    color=world.iloc[:,-1].values,
                    width = 1050,
                    height= 600
                    )
        fig.update_layout(title="Country_wise Distribution of Data Scientists")
        fig.show()    
```

```python
plot_users('Geopandas')
```

```python
plot_users('plotly')
```
