Let's look at the average ages of the participants. It appears that most are between the ages of 22 to 29 years.

```python
x=sorted(data['Q1'].unique())[:-1]
fig = plt.figure(figsize=(10,7))
sns.countplot(data['Q1'][:-2], order=x)
plt.tight_layout()
```

```python
def get_avg_age(r_age):
    ages = r_age.split('-')
    if len(ages)==2:
        return (int(ages[0]) + int(ages[1]))/2
    elif r_age=='70+':
        return 70
    else:
        return r_age
```

```python
def plot_bar(start_index, end_index, title=None, rotation=45, figsize=(10,7)):
# start_index = 221
# end_index = 230
    data_dict = dict()
    for i in range(start_index,end_index):
        desc = ((data.iloc[0][i]).split('-')[-1]).strip()
        column = columns[i]    
        data_dict[desc] = data1[column].count()

    sorted_dict=dict(sorted(data_dict.items(), key=lambda item: item[1], reverse=True))
    y=list(sorted_dict.values())
    x=list(sorted_dict.keys())
    fig = plt.figure(figsize=figsize)
    ax=sns.barplot(x=x, y=y, palette='Set1')
    #plt.xlabel('Language Used')
    plt.title(title)
    plt.xticks(rotation=rotation)

    for p in ax.patches:
       ax.annotate('{}'.format(int(p.get_height())), (p.get_x()+0.1, p.get_height()+50))
    plt.tight_layout()

```
