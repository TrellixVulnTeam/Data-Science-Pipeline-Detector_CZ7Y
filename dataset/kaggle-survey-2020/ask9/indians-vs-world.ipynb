```python
a_20 = df_2020.groupby("In which country do you currently reside?").get_group('India')["What is your age (# years)?"].value_counts()
a_18 = df_2018.groupby("In which country do you currently reside?").get_group('India')["What is your age (# years)?"].value_counts()
a_19 = df_2019.groupby("In which country do you currently reside?").get_group('India')["What is your age (# years)?"].value_counts()
a_17 = df_2017.groupby("Country").get_group('India')["Age"]
da_17 = pd.DataFrame(a_17)
def convert_age(age):
    if 22.0 <= age <= 24.0:
        return "22-24"
    if 18.0 <= age <= 21.0:
        return "18-21"
    if 30.0 <= age <= 34.0:
        return "30-34"
    if 40.0 <= age <= 44.0:
        return "40-44"
    if 60.0 <= age <= 69.0:
        return "60-69"
    if 25.0 <= age <= 29.0:
        return "25-29"
    if 35.0 <= age <= 39.0:
        return "35-39"
    if 55.0 <= age <= 59.0:
        return "55-59"
    if 50.0 <= age <= 54.0:
        return "50-54"
    if 45.0 <= age <= 49.0:
        return "45-49"
    if 70.0 <= age <= 79.0:
        return "70-79"
da_17["Age"] = da_17["Age"].apply(lambda x : convert_age(x))

g_20 = df_2020.groupby("In which country do you currently reside?").get_group('India')["What is your gender? - Selected Choice"].value_counts()
g_19 = df_2019.groupby("In which country do you currently reside?").get_group('India')["What is your gender? - Selected Choice"].value_counts()
g_18 = df_2018.groupby("In which country do you currently reside?").get_group('India')["What is your gender? - Selected Choice"].value_counts()

fig = make_subplots(rows=2, cols=2)
fig.add_trace(
    go.Bar(
    x = a_20.index,
    y = a_20.values,
    marker = dict(color = 'blue',line=dict(color='#000000', width=1.5))
    ,name="2020"
),
    row=1, col=1
)

fig.add_trace(
        go.Bar(
    x = a_19.index,
    y = a_19.values,
    marker = dict(color = 'red',line=dict(color='#000000', width=1.5))
,name="2019"
),
    row=1, col=2
)

fig.add_trace(
     go.Bar(
    x = a_18.index,
    y = a_18.values,
    marker = dict(color = 'yellow',line=dict(color='#000000', width=1.5))
,name="2018"
),
    row=2, col=1
)

fig.add_trace(
     go.Bar(
    x = da_17["Age"].value_counts().index,
    y = da_17["Age"].value_counts().values,
    marker = dict(color = 'pink',line=dict(color='#000000', width=1.5))
,name="2017"
),
    row=2, col=2
)

fig.update_layout(height=1000, width=800, title_text=" Age Distributions")
fig.show()
```
