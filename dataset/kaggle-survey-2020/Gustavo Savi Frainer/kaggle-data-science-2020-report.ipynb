# **1. Overview**

Data Science is a multidisciplinary field focusing to collect information, detect patterns and generate knowledge from data sets.  Considering the huge growth of the area in the last few years, new techniques and tools are being developed every moment.

With the objective of have an overview of the data science market to the aspirings and find out the study options most used, was made an analysis with data from the Kaggle platform. The data used was provided by data science and machine learning surveys of 2018, 2019, and, mainly, 2020.

This work address three major topics: Profile, Technologies, and Market.

## **1.1. Kaggle**
Kaggle is a community for people interested in data science and artificial intelligence with an on-line platform that provide a lot of challenges. Furthermore, it offers courses about basic concepts, public datasets of many subjects, and environments to practice.

## **1.2. The Surveys**
For this analysis, it was used the data from the Kaggle Machine Learning & Data Science Survey of 2018¹, 2019², and, particularly, 2020³. The surveys were realized by the Kaggle and they were shared on the community, by members e-mail list, and on Kaggle's Twitter account.

The surveys had 23,859 (2018), 19,717 (2019), and 20,036 (2020) valid responses from participants in 171 different countries and territories. To protect the participant's privacy, the information was anonymized and free-form responses weren't included.

# **2. Key Results**
* Most professionals are young and do not have many years of experience.
* There is a gender barrier. Even though it's changing, it's still a problem.
* Formal education is really valued. However, online courses are helpful allies to those who want to start or study specific subjects.
* The activities are interdisciplinary. Be curious and open to learning different contexts are important characteristics.
* Python is the most recommended programming language to start.
* It is important to understand the basics of machine learning algorithms. They are used to solve complex problems as well.
* Data science and machine learning are emerging jobs with a big growth rate and great compensation.
* Some countries, like Canada, have a good salary policy. Unfortunately, most countries still have gender-discriminatory compensations. 


# **3. Profile**
The first topic analyzed is the profile of Data Scientists, with information about who they are, where they are from, and their formal education level.

## 3.1. Ages and Genders##

Data Science is a relatively new and rapidly growing field. According to the LinkedIn U.S. Emerging Jobs Report<sup>4</sup>, the data scientist is in the third position of emerging jobs with a growth of 37% in 2020. An interesting point is the Artificial Intelligence, which shares some subjects like machine learning with data science, is in the first position with a growth of 74% in the same year.

The fact of being a new field causes the professionals are, in the mean, younger. In 2020, more than 56% of participants were 29 years old or younger.

```python
def percent(row, total):
    row.Percent = (row.Count/total[row.Year])
    return row
```

```python
ages2020 = kaggle2020[['Age', 'Gender']]

#[Year, Gender]
agesPerYear = pd.concat([
    ages2020.Gender.value_counts(), kaggle2019.Gender.value_counts(),  kaggle2018.Gender.value_counts()], 
    keys=['2020', '2019', '2018'], 
    names=["Year", "Gender"]
).reset_index(name="Count")

agesPerYear = agesPerYear.replace({'Male':'Man', 'Female':'Woman'})

agesPerYearSum = agesPerYear.groupby('Year').Count.sum()

agesPerYear['Percent'] = pd.Series(dtype='float64')
agesPerYear = agesPerYear.apply(percent, args=(agesPerYearSum,), axis=1)
agesPerYear = agesPerYear.sort_values(by=['Year','Percent'], ascending=[True, False])
agesPerYear.loc[(agesPerYear.Year == '2020')][['Gender','Count']].sort_values('Count')
```

```python
sns.set_theme(style="ticks")
f, ax = plt.subplots(figsize=(10,6))
sns.despine(f)

ages2020 = ages2020.sort_values(by=['Age','Gender'])
sns.histplot(
    ages2020,
    x="Age", hue="Gender", hue_order=['Man','Woman', 'Nonbinary','Prefer to self-describe', 'Prefer not to say'],
    multiple="stack",
    palette="muted",
    log_scale=False,
    linewidth= .5
)

plt.ylabel('Participants')

plt.title("Age Groups and Gender")

plt.figtext(0.5,0,
            "Figure 1: Representation of participants age groups and proportion of each gender per age group.",
            **figtext_arg
           )
plt.show()
```

Diversity is very important when looking for new perspectives and to the development of new ideas, even more in a multidisciplinary environment. Unfortunately, the survey audience has a too homogeneous profile concerning gender. In 2020, men represented more than 78% of participants and women only 19%. Meanwhile, nonbinary, self-described, and undeclared together, only 1.8% (see **Figure 2**) .

Although the scenario has evolved in recent years, where the rate of men was over 80%, still be far to have the desirable diversity.

```python
plt.figure(figsize=(12,6))
sns.set_theme(style="whitegrid")

bars = sns.barplot(
    data=agesPerYear,
    x='Year', y='Percent', hue='Gender',
    order=['2018','2019','2020'],
    palette="Set2",
    linewidth= 0
)
sns.despine(left=False)
bars.set_ylim(0, 1)

bars.yaxis.set_major_locator(mticker.FixedLocator(bars.get_yticks().tolist()))
yTicksLabels = ['{:.0%}'.format(x) for x in bars.get_yticks()]
bars.set_yticklabels(yTicksLabels)

plt.ylabel('Proportion')

plt.title('Genders per Year')
plt.figtext(0.5,0,
            "Figure 2: Comparison of genders proportion per year.",
            **figtext_arg
           )

plt.show()
```

## 3.3. Formal Education##
As it is an area that involves a lot of research, formal education in data science ends up being more valued, especially in activities requiring skills usually employed in the academy and scientific methodologies. Also, it is necessary to have a solid background, generally about statistics and IT, to work with the most requested technologies.

**Figure 5** shows the education level reported by participants. Whilst **Figure 6** presents a comparison between role and the level of formal education.

```python
edu = kaggle2020['Education Level'].fillna('I prefer not to answer').value_counts().reset_index(name="Count").rename(columns={"index":"Level"})
edu = edu.rename(index={0:'b',1:'c',2:'a',3:'e',4:'d',5:'g',6:'f'}).sort_index()

edu['Percent'] = edu.Count/(kaggle2020.shape[0])
#edu
```

```python
plt.figure(figsize=(9,7))
sns.set_theme(style="whitegrid")

bars = sns.barplot(x='Count', y='Level', data=edu)

for e in range(len(edu)):
    plt.annotate(
        f"{edu.iloc[e].Percent*100:2.3}%",
        xy=(bars.patches[e].get_width(), bars.patches[e].get_y()+(bars.patches[e].get_height()/2)),
        va='center',
        ha='left'        
    )
plt.xlabel('Participants')
plt.ylabel('')
sns.despine(left=False, bottom=False)
plt.xticks(np.arange(0,8001,500))

xTicksLabels = ['{:.2}'.format(x)+'k' for x in bars.get_xticks()/1000]
bars.set_xticklabels(xTicksLabels)

plt.title('Formal Education')
plt.figtext(0.5,0,
            "Figure 5: Representation of the highest formal education level of participants.",
            **figtext_arg
           )

plt.show()
```

```python
educationPerRole = kaggle2020.loc[:,['Role','Education Level']].dropna()
educationPerRole = educationPerRole.groupby(['Role','Education Level']).size().reset_index(name="Count")

total = educationPerRole.groupby('Role').Count.sum().reset_index(name="Total")

educationPerRole = educationPerRole.join(total.set_index('Role'), on='Role')
educationPerRole['Percent'] = educationPerRole.Count/educationPerRole.Total

educationPerRole_wide = educationPerRole.pivot(index='Role', columns='Education Level', values='Percent')
educationPerRole_wide.rename(columns={'Some college/university study without earning a bachelor’s degree':'Some college/university study without\n earning a bachelor’s degree'}, inplace=True)
columns = [
    'I prefer not to answer',
    'No formal education past high school',
    'Some college/university study without\n earning a bachelor’s degree',
    'Professional degree',
    'Bachelor’s degree',
    'Master’s degree',
    'Doctoral degree'
]

order=[
    'Research Scientist',
    'Statistician',
    'Data Scientist',
    'Machine Learning Engineer', 
    'Other',
    'Data Engineer',
    'Product/Project Manager',
    'Currently not employed',
    'Data Analyst',
    'Software Engineer',
    'Student',
    'DBA/Database Engineer',
    'Business Analyst'
]

educationPerRole_wide = educationPerRole_wide.reindex(order).reindex(columns=columns)
```

```python
plt.figure(figsize=(10,5))

bars = sns.heatmap(educationPerRole_wide, 
            annot=True, 
            fmt='.2%', 
            yticklabels='auto', 
            cmap=sns.color_palette("coolwarm_r", as_cmap=True),
            center=.2
           )
plt.xlabel('')
plt.ylabel('')
plt.title('Role vs Formal Education')
plt.figtext(0.5,-0.6,
            "Figure 6: Heatmap plot comparing the proportion of each formal education level in each role.",
            **figtext_arg
           )

plt.show()
```

# **4. Technologies**
In this topic, the research was referring to the participant's knowledge, experience, and which tools and technics they are using. Many subjects use as a basis multiple-choice questions. For that reason, you may notice cases where the percentage total is over 100%.

## 4.1. Activities##
The first point here shows how interdisciplinary Data Science is. The **Figure 7** represents the important activities at work and evidences the multidisciplinary in the data science domain, where the three top activities involve business, programming, and statistic background.

```python
activities = pd.DataFrame()

for c in kaggle2020.columns:
    if(c.startswith("Q23_")):
        act = kaggle2020[c].value_counts()
        activities = activities.append({'Activity':act.index[0], 'Count':act[0], 'Type':'Use'}, ignore_index=True)
        
activities = activities.sort_values(by='Count', ascending=False)
activities['Percent'] = activities.Count/activities.Count.sum()
activities = activities.reset_index(drop=True)
#activities
```

```python
plt.figure(figsize=(2,8))
sns.set_theme(style="white")

bars = sns.histplot(data=activities,
                    x='Type',
                    hue='Activity',
                    multiple='stack',
                    weights=activities['Count'],
                    legend=False,
                    palette= sns.color_palette("crest", as_cmap=False, n_colors=8)
                   )
sns.despine(left=True, bottom=False)
plt.xlabel('')
plt.ylabel('')
plt.yticks([])
plt.xticks([])

plt.title('Data Science and ML Activities', x=3.5)
plt.figtext(3,0.08,
            "Figure 7: Activities reported as important part of participants role at work.",
            **figtext_arg
           )

bars.patches.reverse()

for a in activities.index:
    plt.annotate(
        " - "+f"{activities.loc[a].Percent*100:2.3}%"+': '+activities.loc[a].Activity,
        xy=(
            bars.patches[a].xy[0]+bars.patches[a].get_width(), 
            bars.patches[a].xy[1]+bars.patches[a].get_height()/2
        ),
        va='center',
        ha='left'        
    )
plt.show()
```

## 4.3. Programming Languages##

The most used languages are Python (in the first place with 77.5%) and R (in the third place with 21.3%) for programming purposes and SQL(in the second place with 37.6%) to query data. It is worth mentioning it's not rare to use more than one language in the same project.

In the most recommended as a first programming language to an aspiring data scientist, Python was chosen by 71.1% of participants. Python is a versatile language, with a lot of tools for data science and a well-established community. It makes that a useful option to start, with so many resources to the learning process and opportunities to apply.

```python
languages = pd.DataFrame()

for c in kaggle2020.columns:
    if(c.startswith("Q7_")):
        lang = kaggle2020[c].value_counts()
        languages = languages.append({'Language':lang.index[0], 'Count':lang[0], 'Type':'Use'}, ignore_index=True)

for c in kaggle2020['Recommended First Language'].value_counts().items():
    languages = languages.append({'Language':c[0], 'Count':c[1], 'Type':'Recommend as First'}, ignore_index=True)

languages = languages.sort_values(by=['Count'], ascending=False)
languages['Percent'] = languages.Count/kaggle2020.shape[0]
```

```python

bars = sns.catplot(x='Language', y='Count', hue='Type', data=languages, kind='bar', height=5, aspect=3, legend_out=False)

b = bars.axes.ravel()[0]
b.patches.sort(key=lambda x: x.get_height(), reverse=True)

for e in range(len(languages)):
    plt.annotate(
        f"{languages.iloc[e].Percent*100:2.1f}%",
        xy=(b.patches[e].get_x()+(b.patches[e].get_width()/2)+0.1, b.patches[e].get_height()),
        va='bottom',
        ha='center',
        rotation=25
    )
plt.xlabel('Programming Language')
plt.ylabel('Participants')

plt.title('Programming Languages')
plt.figtext(0.5,-0.09,
            "Figure 10: Representation of programming languages that are used (multiple choice) and the recommendations for a first programming language to learn.",
            **figtext_arg
           )

plt.show()
```

# **5. Market**

The third major topic in this analysis is the job market. In this part, the dataset was filtered selecting just the participants that are currently working as a Data Engineer, Data Scientist, Data Analyst, Machine Learning Engineer, or Business Analyst.

## 5.1. Yearly Compensation##

The salaries reported are very varied, as can be seen **in Figure 19**. One of the aspects with a higher impact on salary is the location. In the United States of America, 25% earn more than 150 thousand dollars a year. In the Philippines, on the other hand, only 25% earn more than 15 thousand dollars per year.

```python
kaggle2020 = kaggle2020.loc[(kaggle2020.Role.isin(['Data Engineer', 'Data Scientist',
       'Data Analyst', 'Machine Learning Engineer',
       'Business Analyst']))]
```

```python
def compensationCalc(row):
    if pd.isna(row['Yearly Compensation']):
        return row
    vals = row['Yearly Compensation'].replace(',','').split('-')
    if len(vals) == 2:
        vals = (int(vals[0])+int(vals[1]))/2
    else:
        vals = vals[0]
    row['Value'] = float(vals)
    return row
```

```python
compensation = kaggle2020.loc[:,['Gender','Country','Education Level','Yearly Compensation']].dropna(subset=['Yearly Compensation'])
compensation['Yearly Compensation'] = compensation['Yearly Compensation'].replace({'$0-999': '0-999', '> $500,000':'500,000'})
compensation["Value"] = pd.Series(dtype='float64')
compensation = compensation.apply(compensationCalc, axis='columns')
```

```python
yearlyCompensation = compensation[['Yearly Compensation', 'Value']]
yearlyCompensation = yearlyCompensation.sort_values(by="Value")
yearlyCompensation = yearlyCompensation.groupby('Yearly Compensation').Value.value_counts().reset_index(name='Count')
```

```python
plt.figure(figsize=(14,5))
plt.xticks(rotation=90)
sns.barplot(data = yearlyCompensation.sort_values(by='Value'), x='Yearly Compensation', y='Count')

plt.ylabel('Participants')
plt.title('Yearly Compensation')
plt.figtext(0.5,-0.3,
            "Figure 19: The approximate yearly compensation of participants ($USD).",
            **figtext_arg
           )

plt.show()
```

Considering only the top 10 countries in compensation, the median salary is between 90 thousand and 100 thousand dollars a year. The United States of America has the highest median compensation, followed by Switzerland and Israel (see **Figure 20**).

```python
countriesCompensation = compensation[['Country', 'Yearly Compensation', 'Value']]
countriesCompensation = countriesCompensation.sort_values(by="Country")

topCountries = countriesCompensation.groupby(['Country']).Value.agg(['min','mean','median', 'max']).reset_index()
topCountries = topCountries.sort_values(by=['median', 'mean'], ascending=False)[:10]

countriesCompensation = countriesCompensation.loc[(countriesCompensation.Country.isin(topCountries.Country))]
```

```python
plt.figure(figsize=(14,8))

ax = sns.boxplot(data=countriesCompensation, x='Country', y='Value', order=topCountries.Country)
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

plt.xticks(rotation=45, horizontalalignment='right')

yRange = np.append((np.arange(0,200000,20000)),(np.arange(200000,500001,50000)))
plt.yticks(yRange, ['0']+[ '{:4.0f}'.format(x/1000)+'k' for x in yRange[1:]])
plt.xticks(plt.xticks()[0], topCountries.Country.map(lambda c: c if (len(c) <= 20) else c[0:17]+"..."))


plt.title('Top Salaries Countries')
plt.figtext(0.5,-0.125,
            "Figure 20: Yearly compensation of the ten countries with the highest mean salary.",
            **figtext_arg
           )

plt.show()
```

The formal education level is a valued feature and has influence on the yearly compensation. In the top 10 countries, a doctor earns on average 32% more than a master, approximately. Considering all countries, the difference between a doctoral and a master's degree can be over 50%.

```python
educationCompensation = compensation[['Education Level', 'Yearly Compensation', 'Value']]
educationCompensation = educationCompensation.loc[(educationCompensation['Education Level'].isin(['Doctoral degree','Master’s degree','Bachelor’s degree','Professional degree']))]
educationCompensation = educationCompensation.sort_values("Education Level")
```

```python
plt.figure(figsize=(8,8))

ax = sns.boxplot(data=educationCompensation, x='Education Level', y='Value', order=['Doctoral degree','Master’s degree','Bachelor’s degree','Professional degree'])
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

plt.xticks(rotation=45, horizontalalignment='right')

yRange = np.append((np.arange(0,200000,20000)),(np.arange(200000,500001,50000)))
plt.yticks(yRange, ['0']+[ '{:4.0f}'.format(x/1000)+'k' for x in yRange[1:]])

plt.title('Salaries per Formal Education')
plt.figtext(0.5,-0.125,
            "Figure 21: Comparison of yearly compensation per formal education level.",
            **figtext_arg
           )

plt.show()
```

In the first part of this analysis, the Profile, was addressed the gap between the genders and these differences manifest once again in the salaries as shown in **Figure 22**. Considering the case of the United States of America, the median is equal for women and men, but the mean is over 20% more for men. In the case of Germany, the median is, approximately, 36% higher for men than women.

```python
genderCompensations = compensation[['Gender', 'Country', 'Value']].loc[(compensation.Country.isin(topCountries.Country) & compensation.Gender.isin(['Man', 'Woman']))]
others = compensation[['Gender', 'Country', 'Value']].loc[((~compensation.Country.isin(topCountries.Country)) & compensation.Gender.isin(['Man', 'Woman']))]
others.Country = 'Others'

genderCompensations = pd.concat([genderCompensations, others])
#genderCompensations
```

```python
ax = sns.catplot(data=genderCompensations,
            x='Country',
            y='Value',
            hue='Gender',
            order=(np.append(topCountries.Country,['Others'])),
            kind='box',
            height=6,
            aspect=2.8
           )

ax.axes.ravel()[0].xaxis.set_ticks_position('bottom')
ax.axes.ravel()[0].yaxis.set_ticks_position('left')

plt.xticks(rotation=45, horizontalalignment='right')
yRange = np.append((np.arange(0,200000,20000)),(np.arange(200000,500001,50000)))
plt.yticks(yRange, ['0']+[ '{:4.0f}'.format(x/1000)+'k' for x in yRange[1:]])
plt.xticks(plt.xticks()[0], np.append(topCountries.Country.map(lambda c: c if (len(c) <= 20) else c[0:17]+"..."),['Others']))

plt.title('Salaries per Gender')
plt.figtext(0.5,-0.25,
            "Figure 22: Comparison of mean yearly compensation by gender in the countries with a highest mean salary.",
            **figtext_arg
           )

plt.show()
```
