Эти алгоритмы легче изучать, и они дают достаточно хорошие результаты, что делает их наиболее часто используемыми во всех категориях машинного обучения. * Среди тех, у кого опыт машинного обучения менее 3 лет, методы глубокого обучения, такие как сверточные нейронные сети (CNN) и рекуррентные нейронные сети (RNN), используются чаще, чем машины с градиентным ускорением (GBM) и байесовские подходы. ** Это демонстрирует интерес молодого поколения к участию в сообществе глубокого обучения **. Для тех, у кого опыт машинного обучения более 3 лет, использование таких алгоритмов, как GBM и байесовский, намного выше по сравнению с другими алгоритмами.

Как вы реализуете эти алгоритмы, и нужно ли вам создавать их каждый раз, когда вы работаете над проблемой, или они где-то предварительно реализованы, чтобы вы могли с легкостью использовать? Платформа машинного обучения - это решение этих проблем.

```python
ml_fw = multiplechoice_20.iloc[:,65:80]
ml_fw.columns = ['ML exp'] + [x.split('Choice -')[1].split(' (')[0] for x in ml_fw.columns[1:]]
ml_fw = ml_fw.reindex(list(ml_fw['ML exp'].dropna().index))
ml_fw = ml_fw.groupby('ML exp').count().iloc[:-1].reindex(['< 1 years', '1-2 years', '2-3 years', '3-4 years', '4-5 years',
                                                           '5-10 years', '10-20 years', '20 or more years'])
ml_fw = ml_fw.fillna(0).astype('int').iloc[1:]

fig, ax = plt.subplots(figsize=(16,8))
sns.heatmap(ml_fw, annot= True, fmt="d", linewidths=.5, cmap='GnBu')
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.xlabel('ML Frameworks used', fontsize = 'large')
plt.ylabel('ML Experience', fontsize = 'large')
plt.title('ML Experience v/s Frameworks used', fontsize = 'large')
```

* ** Платформа машинного обучения ** относится к интерфейсу, библиотеке или инструменту, которые позволяют разработчикам более легко и быстро создавать модели машинного обучения, не вдаваясь в подробности лежащих в основе алгоритмов. После изучения алгоритмов машинного обучения важно понять и изучить структуры машинного обучения, используемые для создания и развертывания алгоритмов машинного обучения.

В нем реализованы почти все наиболее часто используемые алгоритмы машинного обучения, включая линейную или логистическую регрессию, деревья решений и многие другие. Алгоритмы повышения, такие как Random Forest и Xgboost, также используются многими для создания алгоритмов машинного обучения.

Облачные услуги - это быстрорастущий рынок. Современные технологии, такие как аналитика больших данных, Интернет вещей, искусственный интеллект и даже хостинг веб-приложений и мобильных приложений, требуют больших вычислительных мощностей. Облачные вычисления предлагают предприятиям альтернативу созданию собственной инфраструктуры. ** Благодаря облачным вычислениям любой, кто пользуется Интернетом, может наслаждаться масштабируемой вычислительной мощностью по принципу plug and play **.
```python
cloud_plat = multiplechoice_20.iloc[:,120:130]
cloud_plat.columns = [x.split('Choice -')[1].split(' (')[0] for x in cloud_plat.columns]
cloud_plat = pd.melt(cloud_plat).dropna().groupby('variable')['value'].count().sort_values()[::-1]

fig, ax = plt.subplots(figsize=(15,7))
sns.barplot(x=list(cloud_plat.index), y=list(cloud_plat.values), palette="rocket")
plt.axhline(0, color="k", clip_on=False)
plt.xticks(rotation=45)
plt.xlabel('Cloud Platforms', fontsize = 'large')
plt.ylabel('Count', fontsize = 'large')
plt.title('Cloud Platform Usage', fontsize = 'large')
```

Amazon Web Services позволяет своим подписчикам пользоваться полноценным виртуальным кластером компьютеров в любое время в зависимости от их требований. Вся услуга доступна через Интернет. * GCP предлагает услуги во всех основных сферах, включая вычисления, сети, хранилище, машинное обучение (ML) и Интернет вещей (IoT).
