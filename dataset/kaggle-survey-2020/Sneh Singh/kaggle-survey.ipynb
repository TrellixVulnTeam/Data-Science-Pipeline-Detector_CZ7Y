```python
import pandas as pd 
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.simplefilter('ignore')
```

# Read Data

```python
pd.set_option('display.max_columns',400)
df = pd.read_csv("../input/kaggle-survey-2020/kaggle_survey_2020_responses.csv" )
```

```python
df.head()
```

```python
list(df.columns)
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```

```python
headers = {}
for col in df.columns:
    headers.update({col:f"{col} : {df.loc[0,col]}"})
```

```python
headers
```

```python
df = df.iloc[1:,:].copy()
df = df[['Time from Start to Finish (seconds)','Q1','Q2','Q3','Q4','Q5','Q6','Q7_Part_1',
       'Q8','Q9_Part_1','Q11','Q13','Q14_Part_1']]
```

```python
df['Q1'].unique()
```

```python
plt.figure(figsize=(25,7))
ax = sns.countplot(df['Q1'])
plt.xticks(rotation=90)
plt.xlabel(None)
plt.title(headers['Q1'])
for p in ax.patches:
    ax.annotate("%.1f%%" % (100*float(p.get_height()/df.shape[0])), (p.get_x() + p.get_width() / 2., abs(p.get_height())),
                     ha='center', va='bottom', color='black', xytext=(0, 5),rotation = 'horizontal',
                     textcoords='offset points')
plt.show()
```
