```python
responses = pd.read_csv("/kaggle/input/kaggle-survey-2020/kaggle_survey_2020_responses.csv")
```

```python
responses.head()
```

```python
#Replacing columns_name with 1st row for our convenience in EDA.
responses.columns = responses.iloc[0].values
```

```python
responses.drop([0],inplace=True) #droping 1st row
```

```python
responses.head()
```

```python
!pip install sweetviz
```

```python
import sweetviz
```

```python
sv = sweetviz.analyze(responses.iloc[:,1:],pairwise_analysis='off') 
'''As first column in our dataset is related to time duration and it's not required to analyze it. You can take it if you want.''' 
                                                                    
```

```python
#To view EDA directly in our notebook
sv.show_notebook(w=3200,h=1800)
```

```python
#To create html file of your EDA
sv.show_html("ML_DS_Survey.html")
#you can download it from kaggle output directory
```
