**BASIC INGREDIENTS TO ANALYZE**

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

**DATA LOAD**

```python
df = pd.read_csv('/kaggle/input/kaggle-survey-2020/kaggle_survey_2020_responses.csv')
df.head(10)
```

**DELETE UNUSEFUL COLUMN**

```python
df.drop(['Time from Start to Finish (seconds)'], axis=1, inplace= True)
df
```

> **WORLD AGE DISTRIBUTION**

```python
df_age = df.iloc[1:,0]
age_sorted = df_age.value_counts().sort_index()
X = age_sorted.index
Y = age_sorted.values
title = 'Age Distribution'

plt.title(title)
plt.bar(X,Y)
plt.show()
```

> **WORLD GENDER DISTRIBUTION**

```python
df_gender = df.iloc[1:,1]
gender_sorted = df_gender.value_counts().sort_index()

man_percent = gender_sorted.values[0]/sum(gender_sorted.values)*100
nobinary = gender_sorted.values[1]/sum(gender_sorted.values)*100
Prefer_not_to_say= gender_sorted.values[2]/sum(gender_sorted.values)*100
Prefer_to_self_describe= gender_sorted.values[3]/sum(gender_sorted.values)*100
Woman = gender_sorted.values[4]/sum(gender_sorted.values)*100

X = gender_sorted.index
Y = [man_percent, nobinary, Prefer_not_to_say, Prefer_to_self_describe, Woman]

title = 'World Gender Distribution'


plt.title(title)
gender_dis = plt.bar(X,Y)
gender_dis[0].set_color('lightgreen')
gender_dis[-1].set_color('orange')

plt.show()  
```

```python
plt.pie(Y, labels=X, autopct='%.1f%%')
plt.legend(loc=(1.0, 0.5))
```

> **COUNTRY DISTRIBUTION**

```python
df_country = df.iloc[1:,2]
country_sorted = df_country.value_counts(ascending = True)
top_21 = country_sorted[-21:] ##It is too many.. 

X = country_sorted.index[-21:]
Y = top_21
title = 'Country Distribution'

plt.title(title)
gra = plt.barh(X,Y)
gra[0].set_color('orange')
plt.show()
```

**Korean is 21th. From now on, we will see how the korean data is distributed.**

> **SALARY BY ENTERPRISE SIZE**

```python
#Check the pivot_table
df_uni = df_korea[['Q20','Q24']]
a = pd.pivot_table(df_uni, index=['Q20'], columns=['Q24'],aggfunc=len).fillna(0).astype(int)
a
```

```python
ax = sns.heatmap(a, annot=True, fmt='d',cmap='YlGnBu')
plt.title('Salary by Enterprise Size', fontsize=20)
plt.show()
```
