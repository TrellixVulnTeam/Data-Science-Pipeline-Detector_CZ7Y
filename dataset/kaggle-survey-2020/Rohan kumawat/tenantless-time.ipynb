## The Median outclasses the Mean

Throughout the notebook, we will be using median instead of the mean, so before moving forward we want to showcase one of our favorite topics: MEAN or MEDIAN.

**Mean:** The mean is the central value of a discrete set of numbers: specifically, the sum of values divided by the number of values. Since our childhood, we are always taught that Mean is the best way to deal with the “average” of a given dataset or a problem.

**Median:** The median is a value separating the higher half from the lower half of a data sample. For a dataset, it may be thought of as “the middle” value. Median can also be used as an alternative to the mean.

The **purpose of the mean or median** is to provide a one-number summary of a set of data. The idea of mean is quite tricky. Think of it as a way of summarising a whole lot of numbers, in order to make comparisons. Median is intrinsically understandable. It is the middle number when the values are put in order.

So where am I going with this mean-median thing? When both of their purposes are the same, why am I not using Mean? 

Well, **the advantage of choosing the Median over the Mean** is that it does not get skewed so much by a small proportion of extremely large or small values. It gives us a better idea of a “typical” value. In normal language, a median is not affected by outliers or it is “resistant” because it is not at the mercy of outliers. **Example:** If you have a median of 5 and then add another observation of 80, the median is unlikely to stray far from the 5.

Let’s look at the data that I’ve been exploring throughout the notebook. Let’s calculate the mean and median of the Time column.

```python
print("The mean of the Time column: " + str(round(kag20['Time'].mean(),0)) + " seconds.")
print("The median of the Time column: " + str(round(kag20['Time'].median(),0)) + " seconds.")
```

**Mean: 9156 seconds.**

The number says that on average a respondent took around 9156 seconds or 2.5 hours. Let's leave it here and move forward to the Median.

**Median: 626 seconds.**

This number says that on average a respondent took around 626 seconds or approximately 11 minutes. 

Usually the mean and the median are rather close to each other. If not, the problem is in outliers as I explained. As we can see in the below drawn diagram too:

```python
kag_0_2000 = kag20[kag20["Time"] <= 2000]

kag_Q4 = kag_0_2000.copy(deep=True)
kag_Q4["Q4"] = np.where(kag_Q4["Q4"].isnull(),"Unknown",kag_0_2000["Q4"])
kag_Q4["Q4"].replace(["No formal education past high school", "Some college/university study without earning a bachelor’s degree", "I prefer not to answer"], ["NFE", "NBD", "PNA"], inplace=True)

kag_Q5 = kag_0_2000.copy(deep=True)
kag_Q5["Q5"] = np.where(kag_Q5["Q5"].isnull(),"Unknown",kag_0_2000["Q5"])
```

If you have read carefully you must have noticed that our upper fence is 1979. So, what we did is that we extracted the information from our dataset of the people who filled up the form in less than or equal to 2000 seconds.

Since we have extracted the information, now we will again use the violin plot to gain some particulars.

```python
fig = px.violin(data_frame=kag_0_2000, y="Time", box=True, points='all', template="plotly_white")
fig.update_layout(
    title={
        'text': "Figure 4: Time from start to finish Violin Plot(in secs)",
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'})
fig.show()
```
