# Most used programming language

```python
program_lang = ["Q7_Part_1", "Q7_Part_2", "Q7_Part_3", "Q7_Part_4", "Q7_Part_5", "Q7_Part_6", "Q7_Part_7", "Q7_Part_8",
               "Q7_Part_9", "Q7_Part_10", "Q7_Part_11", "Q7_OTHER"]
lang = {}
for l in program_lang:
    key = survey_data[l].dropna().unique()
    lang[key[0]] = survey_data[l].count()
```

```python
df = pd.DataFrame([lang])
df.plot.bar(figsize=(10,8), title='Programming language usage')
```

> It is found that **Python** is most widely used language
