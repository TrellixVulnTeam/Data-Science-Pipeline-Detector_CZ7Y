```python
import matplotlib.pyplot as plt
import seaborn as sns
import pycountry
```

```python
df = pd.read_csv(r"../input/kaggle-survey-2020/kaggle_survey_2020_responses.csv")
```

```python
df2 = df.drop(index=0,axis=0)
df2
```

## 1. Distribution of Age

```python
sns.set_style("white")
fig = plt.figure(figsize=(15,10))


ax = sns.histplot(df2.Q1,color="#d8bfd8")
ax.set_title("Distribution of age",fontsize=(15))
```

```python
age_df = df2.Q1.value_counts().to_frame(name="Age")
count_sum = sum(age_df.Age)
percentage = age_df.Age / count_sum * 100

percentage
```

### We can know that people between age 21~34 take about 70%

## 2. Distribution of Sex

```python
fig = plt.figure(figsize=(10,10))

sex = df2.Q2.value_counts().to_frame(name="Sex")

ax = plt.pie(sex.Sex,autopct='%1.2f%%',pctdistance = 1.5)

plt.legend(sex.index.to_list())
```
