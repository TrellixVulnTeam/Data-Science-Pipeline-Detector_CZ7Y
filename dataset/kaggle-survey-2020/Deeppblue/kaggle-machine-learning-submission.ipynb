```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import numpy as np
import plotly.express as px
import pandas as pd
surveyXY = pd.read_csv("../input/kaggle-survey-2020/kaggle_survey_2020_responses.csv");
```

```python
surveyXY.head()
```

```python
surveyXY.shape
```

```python
surveyX = surveyXY[['Q1', 'Q2', 'Q3', 'Q4', 'Q5', 'Q6', 'Q7_Part_1', 'Q7_Part_2', 'Q7_Part_3' ]]
surveyX.head()
```

```python
survey = surveyX.iloc[5:,0:]
survey.head()
```

```python
survey["Q1"].value_counts().plot(kind="bar")
plt.title("AGE Groups");
```

```python
survey["Q2"].value_counts().plot(kind="barh")
plt.title("Gender");
```

```python
survey["Q4"].value_counts().plot(kind="bar")
plt.title("Level of Formal Education");
```

```python
survey["Q5"].value_counts().plot(kind="barh")
plt.title("Job Title");
```

```python
import seaborn as sns
sns.catplot(y="Q2", kind="count",
            palette="pastel", edgecolor=".9",
            data= survey);
```

```python
f, ax = plt.subplots(figsize=(8, 6))
sns.countplot(y="Q5", data=survey, color="blue");
ax.set_title('Job titles', size = 14, weight='bold');
```
