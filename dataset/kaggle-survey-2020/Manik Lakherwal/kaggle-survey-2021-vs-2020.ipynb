# Q.What is your age (# years)?

```python
fig, ax = plt.subplots( figsize=(15,10))


ax.tick_params(axis='x', which='major', labelsize=15)
ax.tick_params(axis='y', which='major', labelsize=15)
ax.set_ylabel('Count', weight='semibold', fontsize = 20)
ax.set_xlabel('Age Group', weight='semibold', fontsize = 20)
ax.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.6)



g1=sns.barplot(x=df21['What is your age (# years)?'].value_counts().index,
            y=df21['What is your age (# years)?'].value_counts().values, color = 'orange', alpha = 0.66, label = "2021")


g2=sns.barplot(x=df20['What is your age (# years)?'].value_counts().index,
            y=df20['What is your age (# years)?'].value_counts().values,color='brown', alpha = 0.66, label = "2020")

plt.legend(frameon = False,fontsize=20)
plt.suptitle('Age Distribution of Kaggle Users', fontsize = 20, weight='bold')
plt.show()
```

From the gender distribution graph we can observe :

```python
def Plot_graph2021(df,colname=None, title='X distribution of Kaggle Users ',y_label='',shift1=100,shift2=100):
    sns.set_theme(style="darkgrid")

    f, ax = plt.subplots(figsize=(10, 16))


    sns.set_color_codes("pastel")
    sns.barplot(x=df[colname].value_counts().values,
                y=df[colname].value_counts().index,
                label="Percentage of user", color="g",edgecolor='black', linewidth=0.5, saturation = 0.5)

    ax.legend(ncol=2, loc="lower right", frameon=True)
    ax.set( ylabel="",xlabel="User count")
    sns.despine(left=True, bottom=True)
    
    plt.title(title, weight='bold',fontsize=15)
    plt.xlabel("# Users", weight='semibold',fontsize=15)
    plt.ylabel(y_label, weight='semibold',fontsize=15)

    z=df[colname].value_counts().values
    for n, i in enumerate(df[colname].value_counts().index):    
        ax.text(df[colname].value_counts().values[n]+shift1,n,s=f'{round(z[n]/df.shape[0]*100,1)}%,',color='black') 
        ax.text(df[colname].value_counts().values[n]+shift2,n,s=round(z[n],1),color='black')
```

```python
def Plot_graph2020(df,colname=None, title='X distribution of Kaggle Users',shift1=100,shift2=100,y_label=''):
    sns.set_theme(style="darkgrid")

    f, ax = plt.subplots(figsize=(10, 16))


    sns.set_color_codes("pastel")
    sns.barplot(x=df[colname].value_counts().values,
                y=df[colname].value_counts().index,
                label="Percentage of user", color="brown",edgecolor='black', linewidth=0.5, saturation = 0.5)

    ax.legend(ncol=2, loc="lower right", frameon=True)
    ax.set( ylabel="",xlabel="User count")
    sns.despine(left=True, bottom=True)
    
    plt.title(title, weight='bold',fontsize=15)
    plt.xlabel("# Users", weight='semibold',fontsize=15)
    plt.ylabel(y_label, weight='semibold',fontsize=15)

    z=df[colname].value_counts().values
    for n, i in enumerate(df[colname].value_counts().index):    
        ax.text(df[colname].value_counts().values[n]+shift1,n,s=f'{round(z[n]/df.shape[0]*100,1)}%,',color='black') 
        ax.text(df[colname].value_counts().values[n]+shift2,n,s=f'{round(z[n],1)}',color='black')
```

```python
def plot_donut(df,colname=None, title='X distribution of Kaggle Users ',colors=''):
    plt.figure(figsize=(15,10))
    explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0.2,0.2)
    
    circle = plt.Circle((0,0), 0.7, color='white')
    plt.rcParams['text.color'] = 'black'
    
    plt.pie(df[colname].value_counts().values,
        labels=df21[colname].value_counts().index,
        colors=colors,shadow=True,autopct='%.1f%%' )
    
    p = plt.gcf()
    p.gca().add_artist(circle)
    
    plt.title(title, size=20)
    plt.show()
```
