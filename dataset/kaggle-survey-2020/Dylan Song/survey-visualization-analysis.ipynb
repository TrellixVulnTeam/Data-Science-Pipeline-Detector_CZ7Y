# Education of Data Scientists
In this section, I try to analyze what kind of education data scientists have to see which degree is the most common out of them.

```python
# Filtering the data so that there are only data scientists
data_scientist_responses = all_responses[(all_responses["Q5"] == "Data Scientist")]
data_scientist_responses
```

```python
# Here I put all the different tpyes of degrees from the data into separate variables and count the number of times each one occurs.
bachelors = 0
masters = 0
doctoral = 0
no_education_past_high_school = 0
professional_degree = 0
no_answer = 0
some_college = 0

for _, rows in data_scientist_responses.iterrows():
    if rows["Q4"] == "Bachelor’s degree":
        bachelors += 1
    elif rows["Q4"] == "Master’s degree":
        masters += 1
    elif rows["Q4"] == "Doctoral degree":
        doctoral += 1
    elif rows["Q4"] == "No formal education past high school":
        no_education_past_high_school += 1
    elif rows["Q4"] == "Professional degree":
        professional_degree += 1
    elif rows["Q4"] == "I prefer not to answer":
        no_answer += 1
    elif rows["Q4"] == "Some college/university study without earning a bachelor’s degree":
        some_college += 1
    
(bachelors,
masters,
doctoral,
no_education_past_high_school,
professional_degree,
no_answer,
some_college)

```

```python
# Data Visualization
plt.xkcd(scale=0) # This makes all the graphs look cooler
plt.barh("Bachelor's Degree", bachelors)
plt.barh("Master's Degree", masters)
plt.barh("Doctoral Degree", doctoral)
plt.barh("No education past high school", no_education_past_high_school)
plt.barh("Professional Degree", professional_degree)
plt.barh("I prefer not to answer", no_answer)
plt.barh("Some college/university without earning a bachelor's degree", some_college)

plt.xlabel("Level of Education")
plt.ylabel("Number of Data Scientists")
plt.title("Data Scientist Education")

```

# Analysis
Here are my insights from the data:
* Most data scientists have a master's degree
* A very small amount of data scientists have no education past high school
* Even a bachelor's degree can get you the data scientist job, but a master's degree is preferred!
