**Looks like the highest participation are from Millenials and Gen Z**

```python
colnames = list(df)
print(colnames)
```

```python
df_analysis = df[['Time from Start to Finish (seconds)', 'Q1']]
df['Time from Start to Finish (seconds)'] = df['Time from Start to Finish (seconds)'].astype('int64') 
df['Q1'] = df['Q1'].astype('str') 
df_age_timespent_avg = df.groupby('Q1')['Time from Start to Finish (seconds)'].mean()
df_age_timespent_std = df.groupby('Q1')['Time from Start to Finish (seconds)'].std()
```

```python
df_age_timespent_avg.plot.bar()
```

```python
df_age_timespent_std.plot.bar()
```

```python
df_age_timespent_std.values
```

```python
age_grp = df_age_timespent_avg.index
x_pos = np.arange(len(age_grp))
mn = df_age_timespent_avg.values
std = df_age_timespent_std.values
```

```python
fig, ax = plt.subplots()
ax.bar(x_pos, mn, yerr=std, align='center', alpha=0.5, ecolor='black', capsize=10)
ax.set_ylabel('Time from Start to finish')
ax.set_xticks(x_pos)
ax.set_xticklabels(age_grp)
ax.set_title('Time taken vs Group with Std')
ax.yaxis.grid(True)

```

```python
fig, ax = plt.subplots()
ax.bar(x_pos, mn,  align='center', alpha=0.5, ecolor='black', capsize=10)
ax.set_ylabel('Time from Start to finish')
ax.set_xticks(x_pos)
ax.set_xticklabels(age_grp)
ax.set_title('Time taken to finish the Survey for each group')
ax.yaxis.grid(True)
```

From the graph right above (Time taken to finish the Survey for each group) what we can say that the average time taken by the oldest in age (and the most experienced with life) Kagglers is nearest to the time taken by the youngest Kagglers. And time taken represents how focused and quick these Kagglers were and hence we can say - Kagglers get young as they grow old.
