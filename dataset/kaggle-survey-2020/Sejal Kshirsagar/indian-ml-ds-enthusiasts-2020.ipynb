```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import plotly.graph_objs as go
import plotly.offline as py
import plotly.express as px
import operator
from plotly.subplots import make_subplots
```

```python
data = pd.read_csv('/kaggle/input/kaggle-survey-2020/kaggle_survey_2020_responses.csv')
questions = data.iloc[0, :].T
data = data.iloc[1:, :]

data.rename({'Q1':'AGE GROUP'}, axis = 1, inplace = True) 
data.rename({'Q2':'GENDER'}, axis = 1, inplace = True)
data.rename({'Q3':'COUNTRY'}, axis = 1, inplace = True)
data.rename({'Q4':'EDUCATION'}, axis = 1, inplace = True)
data.rename({'Q5':'TITLE'}, axis = 1, inplace = True)
data.rename({'Q6':'CODING EXP'}, axis = 1, inplace = True)
data.rename({'Q8':'1ST LANG'}, axis = 1, inplace = True)

idata = data[data['COUNTRY'] == 'India']
imale = idata[idata['GENDER'] == 'Man']
ifemale = idata[idata['GENDER'] == 'Woman']
```

<a id="1"></a>
<h2 style='background:#01A9DB; border:0; color:white'><center>Countrywise Distribution</center><h2>

```python
country_wise_distribution = data['COUNTRY'].value_counts()
fig = px.choropleth(country_wise_distribution.values, locations = country_wise_distribution.index,
                    locationmode = 'country names',
                    color = country_wise_distribution.values,
                    labels={'color':'Count', 'locations':'Location'},
                    color_continuous_scale = px.colors.sequential.OrRd)
fig.update_layout(title = "COUNTRYWISE DISTRIBUTION")
fig.show()
```

```python
fig = px.histogram(data, y="COUNTRY", histnorm='percent',template="plotly_dark")
fig.update_layout(xaxis={"ticksuffix":"%"})
fig.update_layout(xaxis_title="COUNT PERCENTAGE")
fig.update_traces(marker_color='rgb(255,128,128)', marker_line_color='rgb(255,128,128)',
                  marker_line_width=1.5, opacity=1)
fig.show()
```

We can see that about **29.2%** of total responses were from people residing in India. 

<a id="2"></a>
<h2 style='background:#01A9DB; border:0; color:white'><center>Genderwise Distribution</center><h2>

```python
counts = data['GENDER'].value_counts(sort=True)
labels = counts.index
values = counts.values
o_pie = go.Pie(labels=labels, values=values, name="Overall", hole=.3, domain={'x': [0,0.46]},
               marker=dict(colors=px.colors.qualitative.Pastel1))

icounts = idata['GENDER'].value_counts(sort=True)
ilabels = icounts.index
ivalues = icounts.values
i_pie = go.Pie(labels=ilabels, values=ivalues, name="India", hole=.3,domain={'x': [0.52,1]},
               marker=dict(colors=px.colors.qualitative.Pastel1))

layout = dict(title = 'GENDER DISTRIBUTION', font=dict(size=12), legend=dict(orientation="h"),
              annotations = [dict(x=0.17, y=0.5, text='Overall', showarrow=False, font=dict(size=20)),
                             dict(x=0.8, y=0.5, text='India', showarrow=False, font=dict(size=20)) ],template="plotly_dark")

fig = dict(data=[o_pie, i_pie], layout=layout)
py.iplot(fig)
```

We can see that overall there is quite large gender gap. In India 22% women participated in the survey which is greater than the overall women participation percentage (19.4%). Whereas 76.8% men from India participated in the survey which is less than overall men participartion percentage (78.8%).
