```python
import numpy as np
from numpy import nan
import pandas as pd
import altair as alt
```

```python
input_files = dict()
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        path_file = os.path.join(dirname, filename)
        input_files[filename[:-4]] = path_file
        print(path_file)
```

```python
responses = pd.read_csv(input_files['kaggle_survey_2020_responses'], low_memory=False)
Q14_question_cols = [col for col in responses.columns if col.startswith('Q14')]
Q14_question_cols
```

```python
df_Q14_question = responses[Q14_question_cols]
df_Q14_question.index.rename('id', inplace=True)
df_Q14_question = df_Q14_question.reset_index()
df_Q14_question
```

```python
df_viz_lib = (df_Q14_question
                .melt(id_vars='id',
                      value_name='library')
                .query('not library.isnull()', engine='python')
                .groupby('library')
                .agg({'id':'size'})
                .query('id > 1')
                .rename(columns={'id':'count'})
                .reset_index())
bars = alt.Chart(df_viz_lib).mark_bar().encode(
    x = 'count',
    y = alt.Y('library', sort='-x'),
    color = 'library'
).properties(
    title = 'Visualization Library Used on a Regular Basis (Data: Kaggle Survey 2020)'
)

text = bars.mark_text(
    align='left',
    baseline='middle',
    dx=3,
    color='black'
).encode(
    text='count'
)

bars + text
```

```python
Q7_question_cols = [col for col in responses.columns if col.startswith('Q7')]
df_Q7_question = responses[Q7_question_cols]
df_Q7_question.index.rename('id', inplace=True)
df_Q7_question = df_Q7_question.reset_index()
df_Q7_question
```

```python
df_language = (df_Q7_question
    .melt(id_vars='id',
          value_name='language')
    .query('not language.isnull()', engine='python')
    .groupby('language')
    .agg({'id':'size'})
    .query('id > 1')
    .rename(columns={'id':'count'})
    .reset_index())

bars = alt.Chart(df_language).mark_bar().encode(
    x = 'count',
    y = alt.Y('language', sort='-x'),
    color = 'language'
).properties(
    title = 'Programming Language Used on a Regular Basis (Data: Kaggle Survey 2020)'
)

text = bars.mark_text(
    align='left',
    baseline='middle',
    dx=3,
    color='black'
).encode(
    text='count'
)

bars + text
```
