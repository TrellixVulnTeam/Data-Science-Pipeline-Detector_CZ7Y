#### Import Necessary Libraries

```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.max_columns',500)
sns.set_style('darkgrid')
%matplotlib inline
```

```python
#Load data
data = pd.read_csv('../input/kaggle-survey-2020/kaggle_survey_2020_responses.csv',low_memory = False)

# drop first row
data.drop(0,inplace = True)

data.head(2)
```

### Kagglers Age Distributions

```python
age = data['Q1']

plt.figure(figsize= (10,5))
sns.countplot(y = age)
plt.title('Age group of Kagglers',fontsize = 20)
plt.xlabel('Total Responders',fontsize = 20)
plt.ylabel('Age Group',fontsize = 20)
```

Most of the Kagglers are of young age between 18 to 39.

### Country of Residence

```python
plt.figure(figsize = (20,5))
sns.countplot(x = data['Q3'], data = data)
plt.title('Country of Residence', fontsize = 20)
plt.xlabel('Countries',fontsize = 20)
plt.ylabel('No of Responders',fontsize = 20)
plt.xticks(rotation = 90)
```

### What is your Gender?

```python
plt.figure(figsize = (20,5))
labels= ['Man','Woman','Prefer not to say','Prefer to self-describe','Nonbinary']
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral','magenta']
gender = data['Q2'].value_counts()
plt.pie(gender,colors = colors, autopct='%1.1f%%', shadow=True,radius = 2,pctdistance = 1.1)
plt.legend(labels,loc = 'best')
plt.show()

```

### What is your education?

```python
edu = data['Q4']

plt.figure(figsize= (10,5))
sns.countplot(y = edu)
plt.title('Highest Level Education',fontsize = 20)
```

### Genderwise Education

```python
plt.figure(figsize= (15,10))
sns.countplot(x = 'Q4',hue ='Q2',data= data)
plt.xticks(rotation = 90)

```

### What is the job profession of kagglers participants

```python
job_count = data['Q5'].value_counts()
total = data.shape[0]

fig,ax = plt.subplots(figsize=[12,10])

ax.bar(job_count.index, job_count, width = 0.5)

for j in job_count.index:
       ax.annotate('{:.2f}%'.format(job_count[j]/total*100),
                  xy=(j, job_count[j]+10))

plt.xticks(rotation=90, fontsize=12)
plt.ylabel('No of Responders',fontsize = 20)
plt.title('Job Profession Distribution', fontsize = 20)
```

It appears quarter of total responders are students followed by data scientist. It would be interesting to know what other job profession represents which is greater than data analyst.
