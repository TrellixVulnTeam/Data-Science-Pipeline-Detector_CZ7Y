```python
#Distribution between age and country the respondents reside?
def country_age_dist(df):
    """
    Visualize distribution of respondents according to age groups 
    across the top 10 countries in terms of responses
    """
    
    # get the list
    ctry_list = df['Q3'].value_counts().head(10).index
    temp = df[df['Q3'].isin(ctry_list)]
    cross = pd.crosstab(temp['Q3'],
                        temp['Q1'],
                        normalize='index')    
    return cross

## visualization
fig, axes = plt.subplots(1,1,figsize=(12,12))
fig.suptitle("Distribution of respondents based on country of residence and age group\n(For top 10 countries each year)", fontsize=20)

sns.heatmap(country_age_dist(kaggle_survey_data20),
            annot=True,
            fmt='.2g',
            linewidths=0.5,
            cmap="Blues",
            cbar=False,
            ax=axes)

axes.set_title('2020 Kaggle Survey')
axes.set_xlabel('Age of respondents(years)')
axes.set_ylabel('Country of residence')
```
