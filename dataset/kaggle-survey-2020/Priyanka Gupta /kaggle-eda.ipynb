```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go

%matplotlib inline
```

```python
di=data[data['In which country do you currently reside?']=='India']
dusa=data[data['In which country do you currently reside?']=='United States of America']
dothers=data[data['In which country do you currently reside?']!='United States of America']
dothers=dothers[dothers['In which country do you currently reside?']!='India']
```

# Below graph shows distribution of gender within specific age range. 

Observation:
From below graph, it can be observed that the most of the Kagglers are between age range 25-29.

```python
plt.figure(figsize=(50,120))
a=sns.countplot(
    data=data,
     y='What is your age (# years)?',hue='What is your gender? - Selected Choice')
plt.ylabel('Age Distribution',fontsize=50)
plt.yticks(size=50)
plt.legend(title="Gender Distribution", fontsize=50, title_fontsize=55)
for p in a.patches:
  val = p.get_width()
  x = p.get_x() + p.get_width() + 10
  y = p.get_y() + p.get_height()/2
  a.annotate(val,(x,y),fontsize=50)
plt.show()
```

#  Programming Experience of Kagglers

```python
ques29='For how many years have you been writing code and/or programming?'

values29=[each for each in data[ques29]]
values29=list(set(values29))

print(values29)
isizes=[len(di[di[ques29]==each]) for each in values29]
usizes=[len(dusa[dusa[ques29]==each]) for each in values29]
osizes=[len(dothers[dothers[ques29]==each]) for each in values29]
fig = go.Figure(data=[
    go.Bar(name='USA', y=usizes, x=values29),
    go.Bar(name='India', y=isizes, x=values29),
    go.Bar(name='Othes', y=osizes, x=values29)
])
fig.update_layout(barmode='group',title_text='Years of Programming Experience ',yaxis=dict(title='Number of People'),xaxis=dict(title='Years'))
fig.show()
```

From the above data, it can be observed that the most of the people having programming experience of 0 - 5 years are from India, whereas people having programming experience of greater than 20 years (20+) are from USA.
