```python
import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('../input/kaggle-survey-2020/kaggle_survey_2020_responses.csv')
data.head()
```

```python
data_r = data[data['Q7_Part_2'].notna()]
ides = []
for i in range(1, 11):
    data_r_ide = data_r[data_r[f'Q9_Part_{i}'].notna()]
    name = data_r_ide[f'Q9_Part_{i}'].iloc[1].strip()
    if name == 'Jupyter (JupyterLab, Jupyter Notebooks, etc)':
        name = 'Jupyter'
    if name == 'Visual Studio Code (VSCode)':
        name = 'VSCode'
    ides.append((len(data_r_ide) * 100 / len(data_r), name))

ides.sort(reverse=True)
names = list(map(lambda x: x[1], ides))
x_pos = np.arange(len(names))
usage = list(map(lambda x: x[0], ides))

plt.figure(figsize=(16, 14))
plt.bar(x_pos, usage, align='center')
plt.xticks(x_pos, names, fontsize=12)
plt.ylabel('% of users', fontsize=14)
plt.title('IDEs used by R programmers', fontsize=14)
plt.show()
```
