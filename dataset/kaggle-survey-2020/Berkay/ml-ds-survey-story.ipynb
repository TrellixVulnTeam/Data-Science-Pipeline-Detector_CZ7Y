# Programming Language


```python
language = df[[i for i in df.columns if 'Q7' in i]]
language_all = pd.Series(dtype='int')
for i in language.columns:
    language_all[language[i].value_counts().index[0]] = language[i].count()

language_all = language_all.sort_values(ascending=True)
language_all
```

```python
programminglan_names = []
programminglan_popularity = []


for x in range(1,13):
    programminglan_names.append(df['Q7_Part_'+str(x)].dropna().unique()[0])
programminglan_names.append(df['Q7_OTHER'].dropna().unique()[0])


for y in range(1,13):
    programminglan_popularity.append(df['Q7_Part_'+str(y)].dropna().count())
programminglan_popularity.append(df['Q7_OTHER'].dropna().count())
                     
    
plt.figure(figsize=(15,12))
sns.set_style('darkgrid')
sns.barplot(y=programminglan_names,x=programminglan_popularity,color='#42b7bd')


ax=plt.gca()
ax.get_children()[0].set_color('#a834a8')
ax.get_children()[2].set_color('#5C5174')
ax.get_children()[1].set_color('#926d88')
plt.title('Most Popular Programming Languages',size=20)
```
