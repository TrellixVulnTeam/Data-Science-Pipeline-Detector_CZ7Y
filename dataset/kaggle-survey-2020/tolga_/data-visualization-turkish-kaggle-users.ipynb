**In this notebook, we examine the survey responses of Turkish Kaggle users.**

## Library

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
survey_data = pd.read_csv('../input/kaggle-survey-2020/kaggle_survey_2020_responses.csv')
```

```python
survey_data.head()
```

## Selection of Turkish Kagglers

```python
survey_tr = survey_data[survey_data["Q3"] =="Turkey"]
print("Row Number : {}".format(survey_tr.shape[0]),5*"-","Column Number : {}".format(survey_tr.shape[1]))
```

```python
survey_tr.head(9)
```

# Age(Years)

```python
age = pd.DataFrame(survey_tr["Q1"].groupby(survey_tr["Q1"]).count())
fig = px.pie(survey_tr,values=survey_tr["Q1"].groupby(survey_tr["Q1"]).count(),names =age.index,title = "Age Distribution of the Participants")
fig.update_traces(textposition='outside', textinfo='percent+label')
fig.show()
```

# Gender 

```python
gender = survey_tr["Q2"].value_counts()
labels = gender.index
fig = go.Figure(data=[go.Pie(labels=labels, values=gender, hole=.3)])
fig.update_traces(textposition='outside', textinfo='percent+label')
fig.show()
```

```python
fig = px.bar(survey_tr["Q1"],color = survey_tr["Q2"],barmode = 'stack',title = "Gender Distribution of Age");fig.show()
```

# Education

```python
edu = survey_tr["Q4"].value_counts()
labels = edu.index
fig = px.bar(edu,title="Level of Education",text=edu)
fig.show()
```

# Title

```python
title = survey_tr["Q5"].value_counts()
labels = title.index
fig = go.Figure(data=[go.Pie(labels=labels, values=title, hole=.2)])
fig.update_traces(textposition='outside', textinfo='percent+label')
fig.update_layout(
    title_text="Dist. of Title")
    # Add annotations in the center of the donut pies.
    
fig.show()
```

```python
fig = px.bar(survey_tr["Q5"],color = survey_tr["Q2"],barmode = 'stack',title = "Gender Distribution of Titles");fig.show()
```

# For how many years have you been writing code and/or programming?

```python
programming = survey_tr["Q6"].fillna("Unknown")
prog = programming.value_counts()
labels = prog.index
fig = px.pie(values=prog,names=labels,title="Year of Writing Code")
fig.update_traces(textposition='outside', textinfo='percent+label');fig.show()
```

# Programming Languages

```python
import plotly.offline as py

#dataPanda = []
#for data in q7_data:
#    trace = go.Bar(y = q7_data[data].value_counts(),name = q7_data[data].value_counts().index[0])
#    a = fig.update_traces(texttemplate='%{text:.2s}', textposition='outside')
#    dataPanda.append([trace,a]) 

# now do the api call
#fig = dict(data=dataPanda)
#py.iplot(fig)
```

```python
q7_data = survey_tr.iloc[:,7:20]
q7 = []
q7index = []
for i in q7_data.columns:
    vc = q7_data[i].value_counts()
    index = vc.index[0]
    q7.append(vc[0]);q7index.append(index)
    
fig = px.bar(q7,title="Regularly Used Programming Languages",labels=q7index)
fig.update_traces(texttemplate=q7index, textposition='outside');fig.update_traces(marker_color='red')
fig.show()
```

```python
q7_title = pd.concat([q7_data,survey_tr["Q5"]],axis=1)
q7_title = q7_title.groupby("Q5").count()
q7_title.columns = ["Python","R","SQL","C","C++","Java","JavaScript","Julia","Swift","Bash","MATLAB","None","Other"]
px.bar(q7_title,barmode="stack",height=800,title="Programming Languages Used by Titles")
```

# What programming language would you recommend an aspiring data scientist to learn first?

```python
first_pl = survey_tr["Q8"].value_counts()
index = first_pl.index
fig = px.bar(survey_tr["Q8"],title="What programming language would you recommend an aspiring data scientist to learn first?",orientation='h')
fig.update_traces( textposition='inside');fig.update_traces(marker_color='blue')
fig.show()
```

# Which IDE's do you use the most?

```python
ide = survey_tr.iloc[:,21:33]
ide["Q9_Part_1"] =ide["Q9_Part_1"][ide["Q9_Part_1"].isna() == False].replace("Jupyter")
ide["Q9_Part_4"] =ide["Q9_Part_4"][ide["Q9_Part_4"].isna() == False].replace("VSCode")
q9 = []
q9_index = []
for i in ide.columns:
    vc = ide[i].value_counts()
    index = vc.index[0]
    q9.append(vc[0]);q9_index.append(index)
    
fig = px.bar(q9,title="Most Used IDE'S",labels=q9_index)
fig.update_traces(texttemplate=q9_index, textposition='outside');fig.update_traces(marker_color="rgba(50, 171, 96, 0.6)")
fig.show()
```

# What type of computing platform do you use most often for your data science projects?

```python
platform = survey_tr["Q11"].value_counts()
index = platform.index
fig = px.pie(values=platform,names=index,title="Most Used Computing Platforms",color_discrete_sequence=px.colors.sequential.RdBu)
fig.update_traces(textposition='inside', textinfo='percent+label',textfont_size=12);fig.show()
```

# What data visualization libraries or tools do you use on a regular basis?

```python
plot = survey_tr.iloc[:,53:65]
q14 = []
q14_index = []
for i in plot.columns:
    v_c = plot[i].value_counts()
    index = v_c.index[0]
    q14.append(v_c[0]);q14_index.append(index)
    
fig = px.bar(q14,title="Most Used Data Visualization Libraries",labels=q14_index)
fig.update_traces(texttemplate=q14_index, textposition='outside');fig.update_traces(marker_color="rgba(50, 171, 96, 0.6)")
fig.show()
```

```python
df = pd.concat([survey_tr["Q5"],plot],axis = 1)
ddf = df.groupby("Q5").count()
ddf.columns = ["Matplotlib","Seaborn","Plotly/Plotly Express","Ggplot/ggplot2","Shiny","D3 js","Altair","Bokeh","Geoplotlib","Leaflet/Folium","None","Other"]
px.bar(ddf,barmode="stack",height=800,title="Data Visualization Libraries Used by Titles")
```
