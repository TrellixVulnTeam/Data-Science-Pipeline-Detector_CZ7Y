```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.gridspec as gridspec
import seaborn as sns


import warnings
warnings.filterwarnings(action='ignore')
plt.rcParams['figure.dpi'] = 200 #high resolution
```

```python
data = pd.read_csv('../input/kaggle-survey-2020/kaggle_survey_2020_responses.csv')
```

```python
print('The dataset has {} rows and {} columns.'.format(len(data), len(data.columns)))
```

### Q1 What is your age

```python
age_q1 = data['Q1'].value_counts()

plt.figure(figsize=(8,2))
sns.barplot(age_q1.index, age_q1.values, alpha=0.5)
plt.ylabel('Number of Q1', fontsize=5)
plt.xlabel('The total age count', fontsize=5)
plt.title('Maximun age count where kaggle is used most')
plt.xticks(rotation=90)
plt.show()
```

* We can see that in the age group between 25-29 are using Kaggle mostly and the second age between 22-24 . 
* The least amount is at the age group of 70+ 

### Q2 What is your gender?

```python
# Data to plot for male or female

gender_q2 = data['Q2'].value_counts()

plt.figure(figsize=(8,2))
sns.barplot(gender_q2.index, gender_q2.values, alpha=0.5)
plt.ylabel('Question Number Q2', fontsize=5)
plt.xlabel('The Gender count', fontsize=5)
plt.title('Type of gender maximum the kaggle are used')
plt.xticks(rotation=90)
plt.show()
```

### Q3 In which country do you currently reside?

#### 1.Argentina

```python
Argentina = d[d['Q3'] == 'Argentina']

fig, ax = plt.subplots(2, 2, figsize=(30, 15))

sns.barplot(Argentina['Q2'].index, Argentina['Q2'].values, alpha=0.5, ax=ax[0,0])
ax[0,0].set_title('Gender count in Argentina')

sns.countplot(Argentina['Q4'].values, alpha=0.5, ax=ax[1,1])
ax[1,1].set_title('Education of all the kagglers in Argentina')
plt.setp( ax[1,1].xaxis.get_majorticklabels(), rotation=70 )

sns.countplot(Argentina['Q5'].values, alpha=0.5, ax=ax[1,0])
ax[1,0].set_title('Position of all the kagglers in Argentina')
plt.setp( ax[1,0].xaxis.get_majorticklabels(), rotation=70 )


sns.countplot(Argentina['Q6'].values, alpha=0.5, ax=ax[0,1])
ax[0,1].set_title('Age group of all the kagglers in Argentina')
plt.setp( ax[0,1].xaxis.get_majorticklabels(), rotation=70 )

fig.suptitle('Over all view of Argentina')
```
