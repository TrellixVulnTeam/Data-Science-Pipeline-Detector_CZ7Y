```python
df=pd.read_csv('/kaggle/input/kaggle-survey-2020/kaggle_survey_2020_responses.csv')
df.head()
```

The columns that are to my interest are Duration (in seconds) and Q5 which is the title most similar to their role.

```python
df1=df[['Time from Start to Finish (seconds)','Q5']]
```

```python
df1
```

```python
df1=df1.dropna()
```

Removing all the rows with NaN values

```python
df1=df1.drop(0)
```

```python
df1
```

```python
z=df1.iloc[1]['Time from Start to Finish (seconds)']
print(type(z))
```

```python
df1['Time from Start to Finish (seconds)'] = df1['Time from Start to Finish (seconds)'].astype(int)
```

```python
df2=df1.groupby('Q5', as_index=False).mean()
```

```python
df2
```

```python
df2['Time In Hour']=df2['Time from Start to Finish (seconds)'].apply(lambda a: a/3600)
```

```python
df2
```

```python
df2=df2.sort_values(by='Time In Hour',ascending=False)
```

```python
df2['Q5']
```

```python
import matplotlib.pyplot as plt  
plt.bar(df2['Time In Hour'],df2['Q5'], color ='maroon')
```
