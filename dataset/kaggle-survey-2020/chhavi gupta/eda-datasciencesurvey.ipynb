#### Q8. Which of the following hosted notebook products is more use by students and professionals?

```python
df_stud = data[data.Q5 == 'Student']
df_prof = data[data.Q5 != 'Student']
q10_columns = [column for column in data.columns if 'Q10' in column]

df_q10 = pd.DataFrame()
print(df_q10)
for column in q10_columns:
    hardware = data[column].dropna().unique()[0]
    df_q10.loc['Student', hardware] = df_stud[column].notnull().sum() / len(df_stud)
    df_q10.loc['Professional', hardware] = df_prof[column].notnull().sum() / len(df_prof)


hardwares = df_q10.columns
fig = go.Figure()
fig.add_trace(go.Bar(
    x=hardwares,
    y=df_q10.loc['Student'],
    name='student',
    marker_color='indianred'
))
fig.add_trace(go.Bar(
    x=hardwares,
    y=df_q10.loc['Professional'],
    name='Professional',
    marker_color='lightsalmon'
))

fig.update_layout(barmode='group', xaxis_tickangle=-45)
fig.show()
```

### Students more used colab notebook while Professionals used Kaggle.
