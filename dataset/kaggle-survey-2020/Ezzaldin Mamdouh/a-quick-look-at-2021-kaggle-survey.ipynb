```python
# getting the number of programming languages and tools that can be used on a regular basis.....
languages = []
for col in ds_df.iloc[:, 7:20].columns:
    lst = ds_df[col].unique().tolist()
    lst.remove(np.NaN)
    languages.append(lst[0])
```

```python
# change column names from column of index 7 to 20..
ds_df.rename(columns = {old: 'use {} on a regular basis'.format(new) for new, old in zip(languages, ds_df.iloc[:, 7:20].columns)}, inplace = True)
```

```python
# label encode the columns of 7 to 20 (np.NaN ----> 0), (val -----> 1)
ds_df = pd.get_dummies(data = ds_df, columns = ds_df.iloc[:, 7:20].columns.tolist())
```

```python
prog_lang_use = ds_df.iloc[:, -13:]
ds_use = pd.concat([ds_df['Q5'],prog_lang_use], axis = 1)
ds_use['total_programming_languages'] = ds_use.iloc[:].sum(axis = 1)
```

```python
ds_coding_count = pd.pivot_table(data = ds_use,
                                 index = 'Q5',
                                 values = ds_use.iloc[:, 1:14].columns.tolist(),
                                 aggfunc = 'sum')
prog_languages = ['Bash', 'C', 'C++', 'Java', 'Javascript', 'Julia', 'Matlab', 'None', 'Other', 'Python', 'R', 'SQL', 'Swift']
ds_coding_count.rename(columns = {old: new for old, new in zip(ds_coding_count.columns, prog_languages)},
                       inplace = True)
ds_coding_count
```

```python
# function to plot pareto chart...
def pareto_chart(series, title=None, xlabel=None, ylabel=None):
    series = series.sort_values(ascending = False)
    x = series.index
    y = series.values
    weights = y / y.sum()
    cumsum = weights.cumsum()
    fig, ax1 = plt.subplots(figsize = (20, 6))
    ax1.bar(x, y, color = '#2680eb')
    if xlabel:
        ax1.set_xlabel(xlabel)
    if ylabel:
        ax1.set_ylabel(ylabel)
    ax2 = ax1.twinx()
    ax2.plot(x, cumsum, '-o', color = '#e68619')
    ax2.set_ylabel('', color='r')
    ax2.tick_params('y', colors='r')
    ax2.set_yticks([])
    #{0:.0%}
    formatted_weights = ['{0:.0%}'.format(x) for x in cumsum]
    for idx, txt in enumerate(formatted_weights):
        ax2.annotate(txt, (x[idx], cumsum[idx]), fontweight = 'bold')
    if title:
        plt.title(title)
    plt.tight_layout()
    plt.show()
```

```python
for idx in ds_coding_count.index:
    z = ds_coding_count.loc[idx]
    pareto_chart(z,
                 title = 'pareto chart of {}s for each programming Language and thier percentages from total {}s'.format(idx, idx), 
                 xlabel = 'Programming Languages')
```
