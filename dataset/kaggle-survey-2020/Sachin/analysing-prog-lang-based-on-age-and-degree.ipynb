# Analyzing the Programing language(Python,R,SQL) based on Age and Degree

```python
df = pd.read_csv('/kaggle/input/kaggle-survey-2020/kaggle_survey_2020_responses.csv')
```

```python
questions = df.iloc[0]
df = df.drop(df.index[0])
df.head()
```

## Generic Analysis of Survey
1. Degree Distribution of survey
2. Age Distribution of survey

### Degree Distribution of survey

```python
df_degree = df.groupby('Q4')['Q4'].count().reset_index(name='deg_count')
df_degree
```

Master degree holders are the most who work around in kaggle competition and datasets followed by Bachelors degree which means specialised degree holders are the major kaggle contributors. The same we can plot on the barplot

```python
import seaborn as sns
from matplotlib import pyplot as plt
plt.figure(figsize=(30,8))
ax0 = sns.barplot(x="Q4", y="deg_count", data=df_degree)
```

### Age Distribution of survey

```python
ds = df['Q1'].value_counts().reset_index()
ds.columns = ['age', 'count']
ds = ds.sort_values(['count'],ascending =False)
plt.figure(figsize=(15,8))
ax1 = sns.barplot(x="age", y="count", data=ds)
```

People from the age group 18-29 forms the major kagglers population, given so many online courses for data science and ML and so many people taking it up can be a possible reason for this age group to be a major user of the platform

### Exploring Python programing language based on Age and Degree

```python
df_age_python = df.groupby(['Q1', 'Q7_Part_1'])['Q7_Part_1'].count().reset_index(name='Age_Python_count')
df_age_python
```

```python
plt.figure(figsize=(15,8))
ax2 = sns.barplot(x="Q1", y="Age_Python_count", data=df_age_python)
```

```python
df_degree_python = df.groupby(['Q4', 'Q7_Part_1'])['Q7_Part_1'].count().reset_index(name='Python_count')
df_degree_python
```

```python
plt.figure(figsize=(30,8))
ax3 = sns.barplot(x="Q4", y="Python_count", data=df_degree_python)
```

Python is the most widely used programing language in the data and ML field. This can possibly be attributed to the fact that since most of the degree holders may have made a career trasition from software development where python is extensively used, also to the fact the specialized degree holders may not be bacheorate, masters or doctorate in statistics and related field where R is intensively used.
