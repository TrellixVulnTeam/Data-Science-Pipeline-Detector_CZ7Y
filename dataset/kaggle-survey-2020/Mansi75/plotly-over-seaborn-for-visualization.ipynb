```python
import pandas as pd
df = pd.read_csv("../input/kaggle-survey-2020/kaggle_survey_2020_responses.csv")
```

Let's make a graph in Seaborn to visualize 'Gender' Count in the dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt
df1_q2 = df['Q2'].value_counts()  
plt.figure(figsize=(10,8))
sns.barplot(df1_q2.index, df1_q2.values, alpha=0.9)
plt.ylabel('Number of Q2', fontsize=8)
plt.xlabel('What is your gender? - Selected Choice', fontsize=7)
plt.xticks(rotation=90)
plt.show();
```

```python

import matplotlib.pyplot as plt
df1_q2 = df['Q1'].value_counts()  
plt.figure(figsize=(10,8))
sns.barplot(df1_q2.index, df1_q2.values, alpha=0.9)
plt.ylabel('Number of Q1', fontsize=8)
plt.xlabel('What is your age (# years)?', fontsize=7)
plt.xticks(rotation=90)
plt.show();
```

Now let visualize the same 'Gender' graph in Plotly.

```python
import plotly.express as px
df['Q2'].isnull().values.any()
gender = df['Q2'].value_counts()
gender = pd.DataFrame(gender)
gender.columns = ['Count']
gender.index.names = ['Gender']
fig = px.bar(gender, y='Count', x=gender.index,text='Count',title='G E N D E R')
fig.update_traces(texttemplate='%{text:.2s}', textposition='outside')
fig.update_layout(uniformtext_minsize=12, uniformtext_mode='hide')
fig.show()
```

When you mouse over to the bars made by plotly, it displays the details about the bar. In this particular example, it is showing Gender and it's count.
This feature is not available in Seaborn.

On the top right of plotly graph, you can see multiple small icons. These are the options/functionalities which make plotly plots more interactive, you save/download the plot as image, can use zoom in and out function not just these but you can play with the axis values too and get a new plot.

On the bottom right you can see option to export the plot to plotly. If you click on that option the plot will open on plotlyâ€™s official website. The idea behind this option is to let user see the values of their chart axes and how it was plotted(e.g. code behind the plotting).

Just hover over the pie chart and you will see the age range as label and the number of count value as the amount of people in the age range. 

```python
import plotly.express as px
age = pd.DataFrame(df["Q1"].groupby(df["Q1"]).count())
fig = px.pie(df,values=df["Q1"].groupby(df["Q1"]).count(),names =age.index,title = "Age Distribution of the Participants")
fig.update_traces(textposition='outside', textinfo='percent+label')
fig.show()
```

As we can see plotly is more beneficial than seaborn for visualization as it has many visualization features that seaborn doesn't have.
