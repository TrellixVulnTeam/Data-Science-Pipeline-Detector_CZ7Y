# Overview of Kaggle Survey Data 2020
bar charts showing the number of each response of each question

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy 
sns.set(rc={'figure.figsize':(12,8)})
```

```python
df = pd.read_csv("../input/kaggle-survey-2020/kaggle_survey_2020_responses.csv", 
                 header=0, skiprows=[1,])
questions = pd.read_csv("../input/kaggle-survey-2020/kaggle_survey_2020_responses.csv", 
                 header=0, nrows=1)
```

```python
def plot_count(column):
    try:
        column = questions.columns[int(column)]
    except:
        None
    question = questions.loc[0, column]
    ax=sns.countplot(data=df, y=column, order=df[column].value_counts().index)
    ax.set_title(question)
    ax.set_ylabel("Answer")
    return ax
```

```python
def plot_count_multi(start, num):
    multi = df.iloc[:,start:start+num]
    counts = pd.DataFrame({'Response':multi.apply(lambda x:x.dropna().unique()[0]), 'Count':multi.count()})
    counts.sort_values(by='Count', ascending=False, inplace=True)
    ax = sns.barplot(data=counts, y='Response', x='Count', order=counts['Response'])
    ax.set_title(questions.iloc[0,start].split(' - Selected')[0])
    return ax
```
