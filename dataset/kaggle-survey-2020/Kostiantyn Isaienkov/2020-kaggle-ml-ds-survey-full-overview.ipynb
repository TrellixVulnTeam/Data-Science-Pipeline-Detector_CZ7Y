```python
import numpy as np
import pandas as pd
        
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objs as go
```

<h2 style='background:blue; border:0; color:white'><center>Q1. What is your age (# years)?</center><h2>

```python
ds = data['Q1'].value_counts().reset_index()
ds.columns = [
    'age', 
    'count'
]
ds = ds.sort_values(['count'])

fig = px.bar(
    ds, 
    x='count', 
    y="age", 
    orientation='h', 
    title='Age distribution', 
    width=800,
    height=600 
)

fig.show()
```
