# **Gender Based Age Distribution**

```python
men = data['Q1'].groupby(data['Q2']).value_counts()['Man'].sort_index()
women = data['Q1'].groupby(data['Q2']).value_counts()['Woman'].sort_index()

plt.style.use('ggplot')
fig, ax = plt.subplots()
ax = fig.add_axes([0,0,1.5,1], title='Gender Based Age Distribution')
w = 0.4
bar1 = np.arange(len(labels))
bar2 = [i+w for i in bar1]

rects1 = ax.bar(bar1, men, width=w, label='Male', edgecolor='black')
rects2 = ax.bar(bar2, women, width=w, label='Female', edgecolor='black')
plt.xticks(bar1+w/2,labels.index)


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
        

autolabel(rects1)
autolabel(rects2)
ax.set_ylim(0, 3500) 
ax.legend(handles=[rects1,rects2])
fig.tight_layout()
plt.show()
```
