```python
#header files.....
#standard ones-
import pandas as pd
import numpy as np
import os
#visualization ones-
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings(action='ignore')
#reading file-
data = pd.read_csv("../input/kaggle-survey-2020/kaggle_survey_2020_responses.csv")
```

```python
questions = data[:1]
data = data[1:]
```

```python
from IPython.core.display import display, HTML
que = ''.join([f'<li>{i}</li>' for i in questions.T[0][1:]])
display(HTML(f'<div style="background: 	#708090; padding: 15px; border-radius: 15px;">  <h3 style="color:#191970; padding:15px"> Survey Questions </h3><ul>{que}</ul> </div>'))
```

### Q1. Age Distribution

```python
fig, ax = plt.subplots(1,1, figsize=(16, 9))
data_q1 = data['Q1'].value_counts().sort_index()
for side in ['top', 'left', 'right']:
    ax.spines[side].set_visible(False)
ax.bar(data_q1.index, data_q1,width = 0.6,
       edgecolor='black', color='#dddddd',
       linewidth=0.9)
for annote in data_q1.index:
    ax.annotate("{}".format(data_q1[annote]),
               xy = (annote, data_q1[annote] + 100),ha ='center',va = 'center',
               fontweight='light', color='#20639B')
ax.set_ylim(0,4300)
ax.set_xticklabels(data_q1.index)
fig.text(0.2,0.9,'Age Distribution',fontsize = 17, fontweight = "bold")
ax.grid(axis = 'y')
plt.show()
```

### Q2. Gender Distribution

```python
from pywaffle import Waffle

data_q2 = data['Q2'].value_counts()
fig = plt.figure(
        FigureClass=Waffle, 
        rows=5,
        columns=15,
        values = data_q2,
        title={'label': 'Gender Distribution', 'fontsize': 14},
        labels=["{}({})".format(a, b) for a, b in zip(data_q2.index, data_q2)],
        legend = {'loc': 'lower left','bbox_to_anchor': (0.05, -0.2),'ncol':7,'frameon':False
                 ,'fontsize':10},
        colors = ('#20639B', '#ED553B', '#3CAEA3', '#F5D55C','#ffc0cb'),
        figsize=(15, 5),
        icons = 'user',
        icon_legend=True)
```
