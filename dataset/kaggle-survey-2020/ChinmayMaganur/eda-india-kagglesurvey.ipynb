```python
df_ind['Q24'].unique()

less=['$0-999']        
ten=['1,000-1,999','2,000-2,999','3,000-3,999','4,000-4,999','5,000-7,499','7,500-9,999']    
tsnd=['10,000-14,999','20,000-24,999','25,000-29,999','40,000-49,999','50,000-59,999','60,000-69,999',
      '70,000-79,999','90,000-99,999']
fty=['100,000-124,999','125,000-149,999','150,000-199,999','200,000-249,999','250,000-299,999','300,000-500,000']
more=['> $500,000']
```

```python
def sal(x):
    if x in less:
        return '<1000'
    elif x in ten:
        return  '1000-9999'
    elif x in tsnd:
        return '1000-99999'
    elif x in fty:
        return '100000-500000'
    elif x in more:
        return '<500000'
    else:
        return np.nan
df_ind['salary']=df_ind['Q24'].apply(lambda x:sal(x))    
df_ind['salary'].value_counts()
```

```python
def col(q):
    col_lst = [name for name in df_ind.columns if q in name]
    colmn = df_ind[col_lst]
    return colmn

q23_5=pd.concat([col('Q23'),col('Q5')],axis=1)
q23_5=q23_5.melt(id_vars = 'Q5', var_name = 'work', value_name = 'response')
q23_5 = pd.DataFrame(q23_5.groupby('Q5')['response'].value_counts())
q23_5.columns= ['n_responses']
q23_5.reset_index(inplace=True)
```

```python
total = pd.DataFrame(df_ind['Q5'].value_counts()).reset_index()
total.columns = ['Q5', 'count']
merge=pd.merge(q23_5,total,on='Q5',how='left')
merge['perc']=round((merge['n_responses']/merge['count'])*100,2)
pvt=pd.pivot_table(merge,index='response',values='perc',columns='Q5')
pvt.index=['Analyze and understand data','Build or run  ML service that improves my product','Data infrastructure ','prototypes','research','Ml improvement and experiment','None','other']

#plt.style.use('fivethirtyeight')
plt.figure(figsize=(20,10))
sns.heatmap(pvt,annot=True,cmap='viridis')

plt.xlabel("Job title",fontdict = {'fontsize':20})
plt.ylabel("Work activities",fontdict={'fontsize':15})
plt.title("Q23: type of work based on Job title ", fontdict = {'fontsize':20}, pad=24);
plt.xticks(fontsize=14, rotation=90)
plt.yticks(fontsize=14)
```
