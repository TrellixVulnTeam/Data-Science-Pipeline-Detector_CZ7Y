```python
import numpy as np
import pandas as pd

from kagglerecipes.wandb_viz import plot_basic_charts

import wandb
wandb.login()
```

```python
df = pd.read_csv('/kaggle/input/kaggle-survey-2020/kaggle_survey_2020_responses.csv')
df_responses = df.drop(0).reset_index(drop=True)
df_responses
```

```python
df_responses.columns.values
```

```python
# response for question number 35 A: Do you use any tools to help manage machine learning experiments?
dfQ35A = df_responses[['Q35_A_Part_1', 'Q35_A_Part_2', 'Q35_A_Part_3', 'Q35_A_Part_4',
       'Q35_A_Part_5', 'Q35_A_Part_6', 'Q35_A_Part_7', 'Q35_A_Part_8',
       'Q35_A_Part_9', 'Q35_A_Part_10', 'Q35_A_OTHER']]

dfQ35A.head()
```

```python
# Response for question number 35 B: In the next 2 years, do you hope to become more familiar with any of these tools for managing ML experiments?
dfQ35B = df_responses[['Q35_B_Part_1', 'Q35_B_Part_2', 'Q35_B_Part_3',
       'Q35_B_Part_4', 'Q35_B_Part_5', 'Q35_B_Part_6', 'Q35_B_Part_7',
       'Q35_B_Part_8', 'Q35_B_Part_9', 'Q35_B_Part_10', 'Q35_B_OTHER']]

dfQ35B.head()
```

```python
TOOLS = ["Neptune.ai", "Weights & Biases", "Comet.ml", "Sacred + Omniboard", "TensorBoard", "Guild.ai", "Polyaxon", "Trains", "Domino Model Monitor", "None", "Other"]
```

# Let's Visualize using W&B

```python
dfQ35A.count().values
```

```python
run = wandb.init(project='KaggleSurvey', job_type="viz")
plot_basic_charts(chart_type='bar', 
                  x_data=TOOLS, 
                  y_data=dfQ35A.count().values, 
                  x_name="ML Management Tools", 
                  y_name="Number of Users", 
                  chart_title="Kaggle Survey 2020: Number of Kaggle users using ML Experiment Tracking tool",
                  wandb_chart_name="ques_35_a")
wandb.finish()
```

```python
dfQ35B.count().values
```

```python
run = wandb.init(project='KaggleSurvey', job_type="viz")
plot_basic_charts(chart_type='bar', 
                  x_data=TOOLS, 
                  y_data=dfQ35B.count().values, 
                  x_name="ML Management Tools", 
                  y_name="Users willing to learn a tool", 
                  chart_title="Kaggle Survey 2020: Which tool you want to learn?",
                  wandb_chart_name="ques_35_b")
wandb.finish()
```
