```python
q2=df.Q2.value_counts().to_dict()
plt.figure(figsize=(8,8))
plt.pie(q2.values(), labels=q2.keys(), autopct='%1.1f%%', shadow=True)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.distplot(df['Time from Start to Finish (seconds)'], color='r', kde=False)
```

```python
q3 = df.Q3.value_counts().head(15).sort_values(ascending=False)

plt.figure(figsize=(12,8))
sns.barplot(q3.index, q3.values)
plt.xticks(rotation=45)
```

```python
q4 = df.Q4.value_counts().head(15).sort_values(ascending=False)

plt.figure(figsize=(12,8))
sns.barplot(q4.index, q4.values, color='orange')
plt.xticks(rotation=45)
```

```python
q6 = df.Q6.value_counts().head(15).sort_values(ascending=False)

plt.figure(figsize=(12,8))
sns.barplot(q6.index, q6.values, color='r')
plt.xticks(rotation=45)
```
