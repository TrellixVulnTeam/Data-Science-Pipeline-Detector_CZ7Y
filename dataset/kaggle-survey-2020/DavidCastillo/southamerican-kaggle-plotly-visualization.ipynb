```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.figure_factory as ff
import plotly.graph_objects as go
import plotly.express as px
from plotly.offline import iplot

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)
```

```python
df = pd.read_csv("../input/kaggle-survey-2020/kaggle_survey_2020_responses.csv")
df.head()
```

```python
questions = df.iloc[0,:].to_dict()
```

<h1> South America countries to analyse </h1>

```python
countries2analyse = [ 'Colombia',"Argentina", "Brazil", "Peru", "Chile" ]
rest_world = df[~df['Q3'].isin(countries2analyse)]
df = df[df['Q3'].isin(countries2analyse)]
df.head()
```

```python
print(f"We have Data for {len(df)} kagglers from South american countries")
```

# Education Q5
Let visualize what is the distribution of the jobs title at south america and the rest of the world


```python
temp_df = df["Q5"].value_counts().reset_index()  
temp_df.columns = ["title", "count"]
temp_df_rest = rest_world["Q5"].value_counts().reset_index()  
temp_df_rest.columns = ["title", "count"]

trace1 = go.Bar(
                x = temp_df['title'],
                y = temp_df['count'], name ="SA",
                marker = dict(color = 'rgb(250,13,92)',
                              line=dict(color='rgb(0,0,0)',width=1.5)),
                text=temp_df['count'], textposition='outside')
trace12 = go.Bar(
                x = temp_df_rest['title'],
                y = -temp_df_rest['count']/max(temp_df_rest['count'])*max(temp_df['count']), name ="World",
                marker = dict(color = 'rgb(12,13,200)',
                              line=dict(color='rgb(0,0,0)',width=1.5)),
                text=temp_df_rest['count'], textposition='outside')
layout = go.Layout(template= "plotly_dark",title = 'JOB TITLES FROM SOUTH AMERICAN KAGGLERS' , xaxis = dict(title = 'Education Level'), yaxis = dict(title = 'Count'), height=650)
fig = go.Figure(data = [trace1,trace12], layout = layout)
fig.update_yaxes(showticklabels=False)
fig.show()
```

# Education Q4

```python
temp_df = df["Q4"].value_counts().reset_index()  
temp_df.columns = ["Degree", "count"]
temp_df_rest = rest_world["Q4"].value_counts().reset_index()  
temp_df_rest.columns = ["Degree", "count"]

trace1 = go.Bar(
                x = temp_df['Degree'],
                y = temp_df['count'], name ="SA",
                marker = dict(color = 'rgb(250,13,92)',
                              line=dict(color='rgb(0,0,0)',width=1.5)),
                text=temp_df['count'], textposition='outside')
trace12 = go.Bar(
                x = temp_df_rest['Degree'],
                y = -temp_df_rest['count']/max(temp_df_rest['count'])*max(temp_df['count']), name ="World",
                marker = dict(color = 'rgb(12,13,200)',
                              line=dict(color='rgb(0,0,0)',width=1.5)),
                text=temp_df_rest['count'], textposition='outside')
layout = go.Layout(template= "plotly_dark",title = 'EDUCATION LEVEL OF SOUTH AMERICAN KAGGLERS' , xaxis = dict(title = 'Education Level'), yaxis = dict(title = 'Count'), height=650)
fig = go.Figure(data = [trace1,trace12], layout = layout)
fig.update_yaxes(showticklabels=False)
fig.show()
```

```python
def pie_plot(cnt_srs, title):
    labels=cnt_srs.index
    values=cnt_srs.values
    trace = go.Pie(labels=labels, 
                   values=values, 
                   title=title, 
                   hoverinfo='percent+value+label', 
                   textinfo='percent',
                   textposition='inside',
                   hole=0.6,
                   showlegend=True,
                   marker=dict(colors=plt.cm.viridis_r(np.linspace(0, 1, 28)),
                               line=dict(color='#000000',
                                         width=2),
                              )
                  )
    
    return trace
```

```python
trace = go.Pie(labels=df['Q4'].value_counts().index, 
               values=df['Q4'].value_counts().values, 
               title="Education Level Distribution", 
               hoverinfo='percent+value+label', 
               textinfo='percent',
               textposition='inside',
               hole=0.6,
               showlegend=True,
               marker=dict(colors=plt.cm.viridis_r(np.linspace(0, 1, 28)),
                           line=dict(color='#000000',
                                     width=2),
                          )
                  )
trace_rest = go.Pie(labels=rest_world['Q4'].value_counts().index, 
               values=rest_world['Q4'].value_counts().values, 
               title="Education Level Distribution", 
               hoverinfo='percent+value+label', 
               textinfo='percent',
               textposition='inside',
               hole=0.6,
               showlegend=True,
               marker=dict(colors=plt.cm.viridis_r(np.linspace(0, 1, 28)),
                           line=dict(color='#000000',
                                     width=2),
                          )
                  )

layout = go.Layout(title="Gender Distribution South America",template= "plotly_dark")
fig=go.Figure(data=[trace], layout = layout)
fig.show()
layout = go.Layout(title="Gender Distribution Rest of the world",template= "plotly_dark")
fig=go.Figure(data=[trace_rest], layout = layout)
fig.show()
```

### We see that more than 78% of the kagglers have a high formal education.
