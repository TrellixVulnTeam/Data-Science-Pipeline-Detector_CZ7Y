## Q12: Which types of specialized hardware do you use on a regular basis? (student vs prefessional)

```python
q12_columns = [column for column in data.columns if 'Q12' in column]

df_q12 = pd.DataFrame()

for column in q12_columns:
    hardware = data[column].dropna().unique()[0]
    df_q12.loc['Student', hardware] = df_stud[column].notnull().sum() / len(df_stud)
    df_q12.loc['Professional', hardware] = df_prof[column].notnull().sum() / len(df_prof)


hardwares = df_q12.columns
fig = go.Figure()
fig.add_trace(go.Bar(
    x=hardwares,
    y=df_q12.loc['Student'],
    name='student',
    marker_color='indianred'
))
fig.add_trace(go.Bar(
    x=hardwares,
    y=df_q12.loc['Professional'],
    name='Professional',
    marker_color='lightsalmon'
))

fig.update_layout(barmode='group', xaxis_tickangle=-45)
fig.show()
```

more professionals are using TPUs than students <br/>
more students are using GPUs than professionals <br/>
more than 30% people in both students and professionals are not any of them
