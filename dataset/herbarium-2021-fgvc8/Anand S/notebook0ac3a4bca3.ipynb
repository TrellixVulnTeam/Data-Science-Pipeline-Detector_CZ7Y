```python
import os
import json
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
#loading json file for train data
data_path_train = "../input/herbarium-2021-fgvc8/train/metadata.json"
with open(data_path_train) as json_file:
    meta_train = json.load(json_file)
#creating seperate df from json
annotations_train =  pd.json_normalize(meta_train ['annotations'])

categories_train =  pd.json_normalize(meta_train ['categories'])
categories_train.rename(columns={'id':'category_id'}, inplace=True)

images_train =  pd.json_normalize(meta_train ['images'])
images_train.rename(columns={'id':'image_id'}, inplace=True)

df_merge = pd.merge(images_train[['image_id','file_name']],annotations_train[['category_id', 'image_id']] , on='image_id')
df_merge = pd.merge(df_merge[['image_id','file_name','category_id']],categories_train[['category_id','name','family','order']] , on='category_id')
#df_merge = df_merge.drop(df_merge[df_merge.order =='Unknown'].index)

df_merge['file_path']="../input/herbarium-2021-fgvc8/train/"+df_merge['file_name']



species= df_merge['category_id'].value_counts()>51
s_list = species.index[species==True].tolist()
df_more= df_merge.loc[df_merge['category_id'].isin(s_list)]

```

```python
def stratified_sample_df(df, col, n_samples):
    n = min(n_samples, df[col].value_counts().min())
    df_ = df.groupby(col).apply(lambda x: x.sample(n))
    df_.index = df_.index.droplevel(0)
    return df_
```

```python
data = stratified_sample_df(df_more, 'category_id', 2000)
```
