```python
DATA_PATH = "../input/herbarium-2021-fgvc8/"
TRAIN_PATH = DATA_PATH + 'train/'
with open(f"{TRAIN_PATH}/metadata.json") as json_file:
    metadata = json.load(json_file)
```

```python
metadata.keys()
```

```python
print(metadata["annotations"][0])
print(metadata["images"][0])
print(metadata["categories"][0])
print(metadata["licenses"][0])
print(metadata["institutions"][0])
```

```python
def plot_image(i, metadata):
    """
    i : indice in the json metadata image list (not the id of the image)
    """
    filename = TRAIN_PATH + metadata["images"][i]['file_name']
    im = imageio.read(filename).get_data(0)
    plt.imshow(im)
    category = metadata["annotations"][i]['category_id']
    category = metadata["categories"][category]
    plt.title(category['name'])
    
plot_image(33, metadata)
```
