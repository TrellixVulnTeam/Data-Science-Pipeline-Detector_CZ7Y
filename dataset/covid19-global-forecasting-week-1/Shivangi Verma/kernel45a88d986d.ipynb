```python
data=df
data.head()
```

```python
data.drop('Province/State',axis=1,inplace=True)
```

```python
data.drop('Country/Region',axis=1,inplace=True)
```

```python
data.drop('Lat',axis=1,inplace=True)
```

```python
data.drop('Long',axis=1,inplace=True)
```

```python
data.drop('Date',axis=1,inplace=True)
```

```python
data.head()
```

```python
x=data
y=data[['Fatalities']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=12212, random_state=1)
```

```python
X_test.count
```

```python
from sklearn.tree import DecisionTreeRegressor  
  
# create a regressor object 
regressor = DecisionTreeRegressor(random_state = 0) 
# fit the regressor with X and Y data 
regressor.fit(x, y) 
```

```python
y_pred = regressor.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.linear_model import LinearRegression
regression_model = LinearRegression()
regression_model.fit(X_train, y_train)
```

```python
# Let us explore the coefficients for each of the independent attributes

for idx, col_name in enumerate(X_train.columns):
    print("The coefficient for {} is {}".format(col_name, regression_model.coef_[0][idx]))
```

```python
intercept = regression_model.intercept_[0]

print("The intercept for our model is {}".format(intercept))
```

```python

regression_model.score(X_test, y_test)
```
