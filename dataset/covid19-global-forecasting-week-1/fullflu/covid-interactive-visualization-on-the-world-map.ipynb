```python
import pandas as pd

import folium
from folium.plugins import HeatMap

from ipywidgets import Play, IntSlider, jslink, HBox, interactive_output
```

# Load data

```python
df_tr = pd.read_csv("/kaggle/input/covid19-global-forecasting-week-1/train.csv")
```

```python
df_tr["date_pd"] = pd.to_datetime(df_tr["Date"])
min_date = df_tr["date_pd"].min()
max_date = df_tr["date_pd"].max()
days =(max_date - min_date).days
```

```python
def f_c(k):
    targetdate = min_date + pd.DateOffset(k)
    base_map = folium.Map(location=[0, 0], zoom_start=2, tiles='cartodbdark_matter')
    heat_data = df_tr.query("date_pd == @targetdate")[["Lat", "Long", "ConfirmedCases"]]
    HeatMap(heat_data).add_to(base_map)
    print(f"Date: {targetdate.date()}")
    display(base_map)
```

```python
play = Play(value=0, min=0, max=days, step=1, interval=1000, description="Press play",)
slider = IntSlider(min=0, max=days)
jslink((play, 'value'), (slider, 'value'))
ui = HBox([play, slider])
out = interactive_output(f_c, {'k': slider})
display(ui, out)
```

```python
def f_f(k):
    targetdate = min_date + pd.DateOffset(k)
    base_map = folium.Map(location=[0, 0], zoom_start=2, tiles='cartodbdark_matter')
    heat_data = df_tr.query("date_pd == @targetdate")[["Lat", "Long", "Fatalities"]]
    HeatMap(heat_data).add_to(base_map)
    print(f"Date: {targetdate.date()}")
    display(base_map)
```

```python
play = Play(value=0, min=0, max=days, step=1, interval=1000, description="Press play",)
slider = IntSlider(min=0, max=days)
jslink((play, 'value'), (slider, 'value'))
ui = HBox([play, slider])
out = interactive_output(f_f, {'k': slider})
display(ui, out)
```
