```python
import pandas as pd 

data = pd.read_csv('../input/covid19-global-forecasting-week-1/train.csv')

data_country = data['Country/Region'].value_counts()

print(data_country)




#print(data.info())
#print(data.head(10))
```

```python
data_unique = data['Country/Region'].unique()

print(data_unique)
```

```python
print(data.info())
# Confirmed Cases 확진환자 
data_unique_date = data['Date'].unique()

print(data_unique_date)
```

```python

condition = (data.Date =='2020-03-01') & (data.ConfirmedCases > 100)

#print (data[condition])
data_0301 = data[condition]

#Province/State	Country/Region 
data_0301_short = data_0301.loc[:,['Province/State','Country/Region','Date','ConfirmedCases']]

# 국가명칭, 날짜, 확진자수

#data_0301.info()
#data_0301.head(10)
data_0301_short.sort_values(by=['ConfirmedCases'], axis=0, ascending=False)

data_0301_short.info()
print(data_0301_short)


```
