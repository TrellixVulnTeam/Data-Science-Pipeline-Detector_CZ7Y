```python
features = pd.read_csv('../input/covid19-global-forecasting-week-1/train.csv')
test_features = pd.read_csv('../input/covid19-global-forecasting-week-1/test.csv')
```

```python
features["Date"] = features["Date"].apply(lambda x: x.replace("-",""))
features["Date"]  = features["Date"].astype(int)
features.head()
```

```python
test_features["Date"] = test_features["Date"].apply(lambda x: x.replace("-",""))
test_features["Date"]  = test_features["Date"].astype(int)
test_features.head()
```

```python
train1_x = features[['Lat','Long','Date']]
train1_y = features[['ConfirmedCases']]
test1_x = test_features[['Lat','Long','Date']]
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(max_depth=200, random_state=1)
model.fit(train1_x, train1_y)
conf_cases = model.predict(test1_x)
```

```python
train2_x = features[['Lat','Long','Date']]
train2_y = features[['Fatalities']]
test2_x = test_features[['Lat','Long','Date']]
```

```python
model = RandomForestClassifier(max_depth=200, random_state=1)
model.fit(train2_x, train2_y)
fatalities = model.predict(test2_x)
```

```python
Sub = pd.read_csv('../input/covid19-global-forecasting-week-1/submission.csv')
output = pd.DataFrame({
    'ForecastId': Sub['ForecastId'],
    'ConfirmedCases': conf_cases,
    'Fatalities': fatalities
})
```

```python
output.head()
```

```python
output.to_csv('submission.csv', index=False)
```
