```python
plt.figure(figsize=(10,5))
sns.set_style("darkgrid")
sns.countplot(data=df_train,x='language',palette = 'viridis_r')
plt.title("Distribution of the training language",fontsize=16)
```

```python
plt.figure(figsize=(10,5))
sns.set_palette('viridis_r')
language=df_train['language'].value_counts()
plt.pie(language,labels=language.index,autopct='%0.2f%%',startangle=270)
plt.title("language Distribution pie chart",fontsize=16)
```

```python
plt.figure(figsize=(10,5))
sns.set_palette("Dark2")
plt.title('language with id',fontsize=16)
sns.lineplot(y = 'language', x='id',data=df_train,linewidth=3)
```

```python
plt.figure(figsize=(15,8))
label=['tamil', 'hindi']
plt.pie(df_train['language'].value_counts().sort_values(), labels=label, explode=[0.15,0.15], 
        autopct='%1.2f%%', startangle=90)
plt.title('Different Types of language')
plt.axis('equal')
```
