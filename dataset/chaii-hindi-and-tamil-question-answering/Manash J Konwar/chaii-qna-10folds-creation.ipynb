# Importing Libraries

```python
import pandas as pd
from sklearn import model_selection
```

# Loading Chaii Train data

```python
train = pd.read_csv("../input/chaii-hindi-and-tamil-question-answering/train.csv")
```

```python
train
```

## Adding Fold Column

```python
train["kfold"] = -1
kf = model_selection.KFold(n_splits=10, shuffle=True, random_state=42)
for fold, (train_indicies, valid_indicies) in enumerate(kf.split(X=train)):
    train.loc[valid_indicies, "kfold"] = fold
```

```python
train.to_csv("chaii_10folds.csv", index=False)
```
