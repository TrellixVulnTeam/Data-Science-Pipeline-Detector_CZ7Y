```python
text = " ".join(paragraphs)
words = text.split(" ")
```

```python
from collections import Counter 
cnt = Counter(words)

cnt.most_common(10)
# print 
```

```python
from spacy.lang.hi import STOP_WORDS as STOP_WORDS_HI
```

```python
#https://colab.research.google.com/github/rahul1990gupta/indic-nlp-datasets/blob/master/examples/Getting_started_with_processing_hindi_text.ipynb#scrollTo=ghnpMGzngcOn

# Let's remove the stop words before printing most common words 

from spacy.lang.hi import Hindi

nlp = Hindi()

doc = nlp(text)

not_stop_words = []
for token in doc:
  if token.is_stop:
    continue
  if token.is_punct or token.text =="|":
    continue 
  not_stop_words.append(token.text)


not_stop_cnt = Counter(not_stop_words)

not_stop_cnt.most_common(10)
```

```python
# Let's render this in wordcloud 
# first import the rquired libraries 
from wordcloud import WordCloud
from spacy.lang.hi import STOP_WORDS as STOP_WORS_HI
import matplotlib.pyplot as plt
```

```python
#https://colab.research.google.com/github/rahul1990gupta/indic-nlp-datasets/blob/master/examples/Getting_started_with_processing_hindi_text.ipynb#scrollTo=ghnpMGzngcOn

wordcloud = WordCloud(
    width=400,
    height=300,
    max_font_size=50, 
    max_words=1000,
    background_color="black",
    colormap='Set3',
    stopwords=STOP_WORDS_HI,
).generate(text)
plt.figure(figsize=(10, 10))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
#https://colab.research.google.com/github/rahul1990gupta/indic-nlp-datasets/blob/master/examples/Getting_started_with_processing_hindi_text.ipynb#scrollTo=ghnpMGzngcOn

# That doesn't look right. We need to provide a custom font file to render it correctly. 
# the issue is highlighted here: https://github.com/amueller/word_cloud/issues/70
import requests
url = "https://hindityping.info/download/assets/Hindi-Fonts-Unicode/gargi.ttf"

r = requests.get(url, allow_redirects=True)
font_path="gargi.ttf"

with open(font_path, "wb") as fw:
  fw.write(r.content)
```

```python
#https://colab.research.google.com/github/rahul1990gupta/indic-nlp-datasets/blob/master/examples/Getting_started_with_processing_hindi_text.ipynb#scrollTo=ghnpMGzngcOn

wordcloud = WordCloud(
    width=400,
    height=300,
    max_font_size=50, 
    max_words=1000,
    background_color="black",
    colormap='Set3',
    stopwords=STOP_WORDS_HI,
    font_path=font_path
).generate(text)
plt.figure(figsize=(10, 10))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
#https://colab.research.google.com/github/rahul1990gupta/indic-nlp-datasets/blob/master/examples/Getting_started_with_processing_hindi_text.ipynb#scrollTo=ghnpMGzngcOn

wordcloud = WordCloud(
    width=400,
    height=300,
    max_font_size=50, 
    max_words=1000,
    background_color="black",
    colormap='Set3',
    stopwords=STOP_WORDS_HI,
    regexp=r"[\u0900-\u097F]+", 
    font_path=font_path
).generate(text)
plt.figure(figsize=(10, 10))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
