```python
import pandas as pd
df = pd.read_csv('../input/chaii-hindi-and-tamil-question-answering/test.csv')
answers = {'id':[],'PredictionString':[]}
for i,row in df.iterrows():
    if row['language']=='hindi':
        sep = 'ред'
    else:
        se= '.'
    ans = row['context'].split(sep)[0]
    answers['id'].append(row['id'])
    answers['PredictionString'].append(ans)
df = pd.DataFrame(answers)
df.to_csv('submission.csv',index=False)
```
