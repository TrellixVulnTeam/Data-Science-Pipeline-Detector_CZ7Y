```python
# Read dataset
train = pd.read_csv("/kaggle/input/chaii-hindi-and-tamil-question-answering/train.csv")
test = pd.read_csv("/kaggle/input/chaii-hindi-and-tamil-question-answering/test.csv")
```

```python
print("Number of samples in tamil language :: ", len(train[train["language"]  == "tamil"]))
print("Number of samples in hindi language :: ", len(train[train["language"]  == "hindi"]))
```

```python
train.head()
```

```python
# Computing character length
train["char_len"] = train["answer_text"].apply(lambda x: len(x))
# Computing word length
train["word_len"] = train["answer_text"].apply(lambda x: len(x.split(" ")))
```

```python
train.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2,2, figsize=(13, 11))

sns.distplot(train[train["language"] == "tamil"]["word_len"], kde=False, ax=axes[0, 0])
axes[0, 0].set_title("Answer word length distribution - tamil")
sns.distplot(train[train["language"] == "tamil"]["char_len"], kde=False, ax=axes[0, 1])
axes[0, 1].set_title("Answer character length distribution - tamil")
sns.distplot(train[train["language"] == "hindi"]["word_len"], kde=False, ax=axes[1, 0])
axes[1, 0].set_title("Answer word length distribution - hindi")
sns.distplot(train[train["language"] == "hindi"]["char_len"], kde=False, ax=axes[1, 1])
axes[1, 1].set_title("Answer character length distribution - hindi")
```

Based on the chart we can say that both hindi and tamil language has similar distribution interms of answer word length. But One catchy insight driven out of this plot can be **"tamil has longer answers than hindi"**

```python
fig, axes = plt.subplots(1,2, figsize=(13, 5))

sns.distplot(train[train["language"] == "tamil"]["answer_start"], kde=False, ax=axes[0])
axes[0].set_title("Answer start position distribution - tamil")
sns.distplot(train[train["language"] == "hindi"]["answer_start"], kde=False, ax=axes[1])
axes[1].set_title("Answer start position distribution - hindi")
```
