## answer mention counts

```python
num_mention_sentence = []
for idx, row in train.iterrows():
    all_sentences = []
    for para in row['context'].split('\n'):
        all_sentences.extend(sentence_split(para, row['language'][:2]))
    mention_count = 0
    for s in all_sentences:
        if s.find(row['answer_text'])>=0:
            mention_count+=1
    num_mention_sentence.append(mention_count)
```

```python
df = pd.Series(num_mention_sentence)
```
