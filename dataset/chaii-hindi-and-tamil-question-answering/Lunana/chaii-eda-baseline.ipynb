14億人近くの人口を抱えるインドは、世界で2番目に人口の多い国です。それでも、ヒンディー語やタミル語などのインドの言語は、ウェブ上では過小評価されています。人気のある自然言語理解（NLU）モデルは、英語と比較してインドの言語でパフォーマンスが低下します。その影響により、インドのユーザー向けのダウンストリームWebアプリケーションでのエクスペリエンスが劣ります。Kaggleコミュニティと新しい機械学習ソリューションからの注目を集めることで、インドのユーザーがWebを最大限に活用できるように支援できます。

質問への回答を予測することは一般的なNLUタスクですが、ヒンディー語とタミル語ではそうではありません。多言語モデリングの現在の進歩には、高品質のデータセットを生成し、モデリングを改善するための集中的な取り組みが必要です。さらに、一般に公開データセットで過小評価されている言語の場合、信頼できる評価を構築するのは難しい場合があります。このコンテストに提供されたデータセット、および参加者によって生成された追加のデータセットが、インドの言語の将来の機械学習を可能にすることを願っています。

このコンテストでのあなたの目標は、ウィキペディアの記事に関する実際の質問への回答を予測することです。質問応答ペアを持つ新しい質問応答データセットであるchaii-1を使用します。データセットは、翻訳を使用せずに収集されたヒンディー語とタミル語を対象としています。それは、ネイティブスピーカーの専門家データアノテーターによって書かれた質問で現実的な情報探索タスクを提供します。構築するためのベースラインモデルと推論コードが提供されます。

タミル語の起源は、他のドラヴィダ語と同様に不明ですが、インドの他の確立された文学言語のほとんどとは異なり、サンスクリット語から独立しています。タミル語はドラヴィダ語の中で最も古い文学を持っていますが（Hart、1975）、言語と文学を正確に年代測定することは困難です。インドまたはスリランカの文学作品は、ヤシの葉の写本（コピーと再コピーの繰り返しを意味する）または口頭での伝達によって保存されていたため、直接の交際は不可能でした。しかし、外部の年代順の記録と内部の言語学的証拠は、現存する最古の作品がおそらく西暦2世紀のいつかで作成されたことを示しています。

タミル語で現存する最も初期のテキストは、古典派の言語を説明する詩学と文法に関する作品であるトルハーッピヤムです。この本の最も古い部分は、紀元前200年頃にさかのぼる可能性があります（Hart、1975）。これらとは別に、今日私たちが持っているタミル語の最も初期の例は、紀元前3世紀の岩の碑文であり、ブラーフミー文字の適応形式であるタミル語-ブラーフ文字で書かれています（Mahadevan、2003）。言語学者は、タミル文学と言語を3つの期間に分類します。古代（西暦前500年から西暦700年）、中世（西暦700年から西暦1500年）、現代（西暦1500年から現在）です。

音声および文学のバリエーション

タミル語は、さまざまな方言に加えて、ケンタミルと呼ばれる形式的または古典的な種類と、コドゥンタミルと呼ばれる口語的な形式との間にかなり鋭いダイグロシアを示します。ダイグロシアは古代からこの言語で存在してきました。初期の寺院の碑文で使用されていた言語は、古典詩の言語とはかなり大きく異なります。結果として、標準のセンタミルは特定の地域のスピーチに基づいていません。これは、タミル語を話すさまざまな地域で書かれた言語をほぼ同じに保つのに役立ちました。

現代では、センタミルは一般的に正式な文章やスピーチで使用されています。それは、例えば、教科書、タミル文学の多く、そして人前で話すことや討論の言語です。しかし、最近では、コドゥンタミルは伝統的にセンタミルの州と見なされてきた地域に侵入しています。たとえば、テレビやラジオでの現代の映画、劇場、人気のある娯楽のほとんどはコドゥンタミルにあり、多くの政治家はそれを使って観客に近づいています。

話された方言はあまり威信を持っていませんでした：タミル人は、文学的なセンタミルの文法規則が神によって定式化されたと信じていたので、それらは唯一の正しいスピーチであると見なされました（たとえば、Kankeyar、1840を参照）。したがって、ほとんどのヨーロッパ言語とは対照的に、タミル語はその歴史の多くで標準的な話し言葉を持っていませんでした。しかし、現代では、コドゥンタミルの使用の増加により、非公式の「標準的な」方言が出現しています。インドでは、「標準的な」コドゥンタミルは、1つの方言ではなく、「教育を受けた非バラモン語」に基づいていますが（Schiffman、1998）、タンジャヴルとマドゥライの方言の影響を大きく受けています。スリランカでは、標準はジャフナの方言に基づいています。

方言

タミル語の方言は、古いタミル語から進化する際に異なる音韻変化と音の変化を経験したという事実によって、主に互いに区別されます。したがって、「ここ」の単語-Centamil（古典的な品種）のingu-はコインバトールのKongu方言のinge、Thanjavurの方言のinga、Tirunelveliの方言のingane、Ramanathapuramの方言のinguttu、ingaleに進化しましたスリランカのさまざまな北方言のインゲードといくつかの方言のインガイ。

ほとんどのタミル語方言は語彙にそれほど大きな違いはありませんが、いくつかの例外があります。スリランカで話されている方言は、インドでは日常的に使用されていない多くの単語を保持しており、他の多くの単語の使用方法は少し異なります。パラカッドのアヤールの方言には、多数のマラヤーラム語の外来語があり、マラヤーラム語の構文の影響も受けており、明確なマラヤーラム語のアクセントもあります。最後に、前者はタミル語アヤールのグループによって話され、後者の2つは11世紀にカルナタカに移住したVaishnavitesによって話された、Sanketi、Hebbar、およびMandyamの方言は、9番目に設計されたタミル語の特別な形式であるVaishnavaparibasaiの多くの機能を保持しています。バンガロアには独自のタミル語もあり、主に母国語がタミル語ではなく、カンナダ語やヒンディー語からの言葉を吹き込んでいる人々によって話されています。

タミル語の方言は、地域とコミュニティの両方によって異なります。いくつかのカーストには独自の方言があり、そのカーストのほとんどのメンバーは、出身地に関係なく伝統的に使用していました。これらの違いのいくつかは、反カースト運動の結果として近年消え始めていますが、多くの痕跡が残っており、スピーチによって人のカーストを識別することがしばしば可能です。

```python
max_length = 384 # The maximum length of a feature (question and context)
doc_stride = 128 # The authorized overlap between two part of the context when splitting it is needed.
```

```python
pad_on_right = tokenizer.padding_side == "right"
```

```python
def prepare_train_features(examples):
    examples["question"] = [q.lstrip() for q in examples["question"]]

    tokenized_examples = tokenizer(
        examples["question" if pad_on_right else "context"],
        examples["context" if pad_on_right else "question"],
        truncation="only_second" if pad_on_right else "only_first",
        max_length=max_length,
        stride=doc_stride,
        return_overflowing_tokens=True,
        return_offsets_mapping=True,
        padding="max_length",
    )

    sample_mapping = tokenized_examples.pop("overflow_to_sample_mapping")
    offset_mapping = tokenized_examples.pop("offset_mapping")

    tokenized_examples["start_positions"] = []
    tokenized_examples["end_positions"] = []

    for i, offsets in enumerate(offset_mapping):
        input_ids = tokenized_examples["input_ids"][i]
        cls_index = input_ids.index(tokenizer.cls_token_id)

        sequence_ids = tokenized_examples.sequence_ids(i)

        sample_index = sample_mapping[i]
        answers = examples["answers"][sample_index]
        if len(answers["answer_start"]) == 0:
            tokenized_examples["start_positions"].append(cls_index)
            tokenized_examples["end_positions"].append(cls_index)
        else:
            start_char = answers["answer_start"][0]
            end_char = start_char + len(answers["text"][0])

            token_start_index = 0
            while sequence_ids[token_start_index] != (1 if pad_on_right else 0):
                token_start_index += 1

            token_end_index = len(input_ids) - 1
            while sequence_ids[token_end_index] != (1 if pad_on_right else 0):
                token_end_index -= 1

            if not (offsets[token_start_index][0] <= start_char and offsets[token_end_index][1] >= end_char):
                tokenized_examples["start_positions"].append(cls_index)
                tokenized_examples["end_positions"].append(cls_index)
            else:
                while token_start_index < len(offsets) and offsets[token_start_index][0] <= start_char:
                    token_start_index += 1
                tokenized_examples["start_positions"].append(token_start_index - 1)
                while offsets[token_end_index][1] >= end_char:
                    token_end_index -= 1
                tokenized_examples["end_positions"].append(token_end_index + 1)

    return tokenized_examples
```

```python
def prepare_validation_features(examples):
    examples["question"] = [q.lstrip() for q in examples["question"]]

    tokenized_examples = tokenizer(
        examples["question" if pad_on_right else "context"],
        examples["context" if pad_on_right else "question"],
        truncation="only_second" if pad_on_right else "only_first",
        max_length=max_length,
        stride=doc_stride,
        return_overflowing_tokens=True,
        return_offsets_mapping=True,
        padding="max_length",
    )

    sample_mapping = tokenized_examples.pop("overflow_to_sample_mapping")

    tokenized_examples["example_id"] = []

    for i in range(len(tokenized_examples["input_ids"])):
        sequence_ids = tokenized_examples.sequence_ids(i)
        context_index = 1 if pad_on_right else 0

        sample_index = sample_mapping[i]
        tokenized_examples["example_id"].append(examples["id"][sample_index])

        tokenized_examples["offset_mapping"][i] = [
            (o if sequence_ids[k] == context_index else None)
            for k, o in enumerate(tokenized_examples["offset_mapping"][i])
        ]

    return tokenized_examples
```

```python
import collections

examples = valid_dataset
features = validation_features

example_id_to_index = {k: i for i, k in enumerate(examples["id"])}
features_per_example = collections.defaultdict(list)
for i, feature in enumerate(features):
    features_per_example[example_id_to_index[feature["example_id"]]].append(i)
```

```python
from tqdm.auto import tqdm

def postprocess_qa_predictions(examples, features, raw_predictions, n_best_size = 20, max_answer_length = 30):
    all_start_logits, all_end_logits = raw_predictions
    example_id_to_index = {k: i for i, k in enumerate(examples["id"])}
    features_per_example = collections.defaultdict(list)
    for i, feature in enumerate(features):
        features_per_example[example_id_to_index[feature["example_id"]]].append(i)

    predictions = collections.OrderedDict()

    print(f"Post-processing {len(examples)} example predictions split into {len(features)} features.")

    for example_index, example in enumerate(tqdm(examples)):
        feature_indices = features_per_example[example_index]

        min_null_score = None 
        valid_answers = []
        
        context = example["context"]
        for feature_index in feature_indices:
            start_logits = all_start_logits[feature_index]
            end_logits = all_end_logits[feature_index]
            offset_mapping = features[feature_index]["offset_mapping"]

            cls_index = features[feature_index]["input_ids"].index(tokenizer.cls_token_id)
            feature_null_score = start_logits[cls_index] + end_logits[cls_index]
            if min_null_score is None or min_null_score < feature_null_score:
                min_null_score = feature_null_score

            start_indexes = np.argsort(start_logits)[-1 : -n_best_size - 1 : -1].tolist()
            end_indexes = np.argsort(end_logits)[-1 : -n_best_size - 1 : -1].tolist()
            for start_index in start_indexes:
                for end_index in end_indexes:
                    if (
                        start_index >= len(offset_mapping)
                        or end_index >= len(offset_mapping)
                        or offset_mapping[start_index] is None
                        or offset_mapping[end_index] is None
                    ):
                        continue
                    if end_index < start_index or end_index - start_index + 1 > max_answer_length:
                        continue

                    start_char = offset_mapping[start_index][0]
                    end_char = offset_mapping[end_index][1]
                    valid_answers.append(
                        {
                            "score": start_logits[start_index] + end_logits[end_index],
                            "text": context[start_char: end_char]
                        }
                    )
        
        if len(valid_answers) > 0:
            best_answer = sorted(valid_answers, key=lambda x: x["score"], reverse=True)[0]
        else:
            best_answer = {"text": "", "score": 0.0}
        
        predictions[example["id"]] = best_answer["text"]

    return predictions
```

```python
def jaccard(row): 
    str1 = row[0]
    str2 = row[1]
    a = set(str1.lower().split()) 
    b = set(str2.lower().split())
    c = a.intersection(b)
    return float(len(c)) / (len(a) + len(b) - len(c))
```
