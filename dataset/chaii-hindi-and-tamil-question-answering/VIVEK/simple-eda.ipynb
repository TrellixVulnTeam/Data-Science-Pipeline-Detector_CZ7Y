## Let's analyze the language distributions 

```python
print(train.language.value_counts(), end = "\n\n")
sns.countplot("language" , data = train)
```

#### **Observation** : The hindi text data points are almost twice of the tamil text data points 

```python
print(test.language.value_counts(), end = "\n\n")
sns.countplot("language" , data = test)
```

## Lets analyze Hindi and Tamil text differently  

```python
train_h = train [ train["language"] == "hindi" ]
train_t = train [ train["language"] == "tamil" ]
```

### Shortest questions( based on number of words )

```python
print("Hindi: \n\n")
display(sorted(train_h["question"].tolist() , key = lambda x : len(x.split(sep = " ")) )[:3])
print("\n\nTamil: \n\n")
display(sorted(train_t["question"].tolist() , key = lambda x : len(x.split(sep = " ")) )[:3])
```

### Longest questions( based on number of words )

```python
print("Hindi: \n\n")
display(sorted(train_h["question"].tolist() , key = lambda x : len(x.split(sep = " ")) )[-3:])
print("\n\nTamil: \n\n")
display(sorted(train_t["question"].tolist() , key = lambda x : len(x.split(sep = " ")) )[-3:])
```

### Distribution of questions length 

```python
plt.figure(figsize = ( 15, 8))
sns.displot( data =  pd.DataFrame({ "Hindi" : train_h["question"].map(lambda x : len(x.split(sep = " "))) , 
                           "Tamil" : train_t["question"].map(lambda x : len(x.split(sep = " ")) ) }) ) 
```

####  **Observations:** Most of the Hindi questions are longer than tamil questions 

> NOTE: this EDA is to look at the data from  high level perspective and not going deep into language understanding

```python
plt.figure(figsize = ( 15, 8))
sns.displot( data =  pd.DataFrame({ "Hindi" : train_h["context"].map(lambda x : len(x.split(sep = " "))) , 
                           "Tamil" : train_t["context"].map(lambda x : len(x.split(sep = " ")) ) }) ) 
```

#### **Observation:** the distribution of hindi and tamil context lengths is similler

### Distribution of hindi and tamil answers length 

```python
plt.figure(figsize = ( 15, 8))
sns.displot( data =  pd.DataFrame({ "Hindi" : train_h["answer_text"].map(lambda x : len(x.split(sep = " "))) , 
                           "Tamil" : train_t["answer_text"].map(lambda x : len(x.split(sep = " ")) ) }) ) 
```

#### **Observations:**

* Most of the answers contains less than 10 words.
