```python
tamil_text = " ".join(train[train["language"]=="tamil"]["question"])
hindi_text = " ".join(train[train["language"]=="hindi"]["question"])
```

```python
from spacy.lang.hi import Hindi
from spacy.lang.ta import Tamil
from spacy.lang.hi import STOP_WORDS as hindi_stopwords
from spacy.lang.ta import STOP_WORDS as tamil_stopwords
from collections import Counter
```

**For Hindi**

```python
hindi_nlp = Hindi()
hindi_doc = hindi_nlp(hindi_text)
hindi_tokens = set([token.text for token in hindi_doc])
hindi_tokens_counter = Counter(hindi_tokens)
```

**For Tamil**

```python
tamil_nlp = Tamil()
tamil_doc = hindi_nlp(tamil_text)
tamil_tokens = set([token.text for token in tamil_doc])
tamil_tokens_counter = Counter(tamil_tokens)
```

```python
from wordcloud import WordCloud
```

**Hindi WordCloud**

```python
wordCloudHindi = WordCloud(font_path="./Devanagari/gargi.ttf",width=500,collocations=True,
                      height=500,min_font_size=8,stopwords=hindi_stopwords,background_color="white").generate_from_frequencies(hindi_tokens_counter)
plt.figure(figsize=(10, 10))
plt.imshow(wordCloudHindi)
plt.axis("off")
plt.show()
```

**Tamil WordCloud**

```python
wordCloudHindi = WordCloud(font_path="./Tamil/Samyak-Tamil.ttf",width=500,collocations=True,
                      height=500,min_font_size=8,stopwords=hindi_stopwords,background_color="white").generate_from_frequencies(tamil_tokens_counter)
plt.figure(figsize=(10, 10))
plt.imshow(wordCloudHindi)
plt.axis("off")
plt.show()
```
