```python
import plotly.graph_objs as go
import plotly.figure_factory as ff
import plotly.express as px
from plotly.subplots import make_subplots
from plotly.offline import iplot
```

```python
train=pd.read_csv("../input/chaii-hindi-and-tamil-question-answering/train.csv")
test=pd.read_csv("../input/chaii-hindi-and-tamil-question-answering/test.csv")
sub=pd.read_csv("../input/chaii-hindi-and-tamil-question-answering/sample_submission.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
print("Number of training samples: ", len(train))
print("Number of test samples: ", len(test))
print("\nTraining data language count:\n",train["language"].value_counts())
plt.figure(figsize=(8, 5))
sns.countplot(data=train, x="language", color='#88fa45')
plt.show()
```

```python
sns.scatterplot(x="language", y="id", data=train)
plt.show()
```

```python
hindi = train[train['language']=='hindi']['context'].str.len()
tamil = train[train['language']=='tamil']['context'].str.len()
fig = ff.create_distplot([hindi, tamil], ['Hindi', 'Tamil'])
fig.update_layout(height=400, width=800, title_text="Character Count per Language")
iplot(fig)
```

```python
hindi = train[train['language']=='hindi']['context'].str.split().map(lambda x: len(x))
tamil = train[train['language']=='tamil']['context'].str.split().map(lambda x: len(x))
fig = ff.create_distplot([hindi, tamil], ['Hindi', 'Tamil'])
fig.update_layout(height=400, width=800, title_text="Character Count per Language")
iplot(fig)
```

```python
hindi = train[train['language']=='hindi']['context'].str.split().map(lambda x: [len(j) for j in x]).map(lambda x: np.mean(x)).to_list()
tamil = train[train['language']=='tamil']['context'].str.split().map(lambda x: [len(j) for j in x]).map(lambda x: np.mean(x)).to_list()

fig = ff.create_distplot([hindi, tamil], ['Hindi', 'Tamil'])
fig.update_layout(height=500, width=800, title_text="Average Word Length Distribution per Language")
iplot(fig)
```
