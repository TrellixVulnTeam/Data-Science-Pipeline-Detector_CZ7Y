```python
from googletrans import Translator
import time
```

```python
INPUT_DIR = '../input/chaii-hindi-and-tamil-question-answering'
OUTPUT_DIR = '../output/'
```

```python
df_train = pd.read_csv(os.path.join(INPUT_DIR, 'train.csv'))
df_train.columns
```

# WARNING
- If you throw a lot of translation requests at once, you will be blocked from the Google Translator for a while. In this code, only the first 5 records are processed, and two-second sleeps are intentionally implemented.

```python
translator = Translator()
translator.raise_Exception = True
def translate(src_text, src, dest='en'):
    src = 'ta' if src=='tamil' else 'hi'
    src_text = str(src_text)
    if src_text is None or src_text == '':
        return ''
    
    translation = translator.translate(src_text, src=src, dest=dest)
    time.sleep(2) 
    return translation.text
```

```python
df_train['question_en'] = None
df_train['answer_text_en'] = None

df_train['question_en'][:5] = [translate(text, src=lang) for text, lang in zip(df_train['question'][:5], df_train['language'][:5])]
df_train['answer_text_en'][:5] = [translate(text, src=lang) for text, lang in zip(df_train['answer_text'][:5], df_train['language'][:5])]
df_train.head()
```
