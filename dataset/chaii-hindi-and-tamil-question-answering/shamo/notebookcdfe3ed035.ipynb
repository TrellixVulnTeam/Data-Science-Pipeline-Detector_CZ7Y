自然言語処理のためにひとまず日本語に翻訳する
Google翻訳のAPIを利用してヒンディー語→日本語に翻訳

```python
pip install googletrans==4.0.0-rc1
```

```python
from googletrans import Translator

translator = Translator()

txt = "ज्वाला गुट्टा (जन्म: 7 सितंबर 1983; वर्धा, महाराष्ट्र) एक भारतीय बैडमिंटन खिलाडी हैं।"

#　日本語へ翻訳
trans_ja = translator.translate(txt, src = "hi", dest = "ja")
print(trans_ja.text)
```
