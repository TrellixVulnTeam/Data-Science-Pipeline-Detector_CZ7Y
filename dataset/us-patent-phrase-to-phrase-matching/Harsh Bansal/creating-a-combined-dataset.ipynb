```python
import pandas as pd
import numpy as np
```

```python
train=pd.read_csv("../input/us-patent-phrase-to-phrase-matching/train.csv")
test=pd.read_csv("../input/us-patent-phrase-to-phrase-matching/test.csv")
titles=pd.read_csv("../input/cpc-codes/titles.csv")
```

```python
sample1=train.merge(titles,left_on="context",right_on="code")
sample1=sample1.merge(titles,left_on="section",right_on="code")
sample1=sample1.dropna(axis=1)
sample1["combined_text"]=sample1["anchor"]+'[SEP]'+sample1["target"]+'[SEP]'+sample1["title_y"]+'[SEP]'+sample1["title_x"]
sample1["combined_text"]=sample1["combined_text"].str.lower()
train_df=sample1[["id","anchor","target","score","context","combined_text"]]
```

```python
train_df.to_csv("train_df.csv",index=False)
```

```python
test_=test.merge(titles,left_on="context",right_on="code")
test_=test_.merge(titles,left_on="section",right_on="code")
test_=test_.dropna(axis=1)
test_["combined_text"]=test_["anchor"]+'[SEP]'+test_["target"]+'[SEP]'+test_["title_y"]+'[SEP]'+test_["title_x"]
test_["combined_text"]=test_["combined_text"].str.lower()
test_df=test_[["id","anchor","target","context","combined_text"]]
```

```python
test_df.to_csv("test_df.csv",index=False)
```
