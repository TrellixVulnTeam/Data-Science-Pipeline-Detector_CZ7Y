```python
import en_core_web_lg
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.svm import SVC
clf = make_pipeline(StandardScaler(), SVC(gamma='auto'))


nlp = en_core_web_lg.load()



df = pd.read_csv('../input/us-patent-phrase-to-phrase-matching/train.csv')
l=len(df)
pp=10000
b=[[]] *pp
c=[[]] *10000
v=[[]]*110000


ar=[]         
for n in range(pp):
    
   
    m = df.loc[n].at["anchor"]
    k = df.loc[n].at["target"]
    
    search_doc = nlp(m)
    
    main_doc = nlp(k)
    e=[main_doc.similarity(search_doc)]
    
    a=np.concatenate((ar, e)) 
      
    
    b[n] = a.tolist() 
      
arrayOfArrays = np.array(b)
arrayOfArrays = arrayOfArrays.tolist()      

x = np.array(arrayOfArrays)       

X = np.array(x)


arr=[]
for n in range(pp):
    
   
    mm = df.loc[n].at["score"]
    
    if mm==0:
        ff=[1]
        arr=np.append(arr,ff)
    if mm==0.25:
        ff=[0]
        arr=np.append(arr,ff)
    if mm==0.5:
        ff=[0]
        arr=np.append(arr,ff)
    if mm==0.75:
        ff=[0]
        arr=np.append(arr,ff)
    if mm==1:
        ff=[0]
        arr=np.append(arr,ff)
    
       
y=arr.tolist()
y = np.array(y)  


# most important SVR parameter is Kernel type. It can be #linear,polynomial or gaussian SVR. We have a non-linear condition #so we can select polynomial or gaussian but here we select RBF(a #gaussian type) kernel.

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=109) # 70% training and 30% test

clf.fit(X_train, y_train)
y = y.ravel()
y_pred2 = clf.predict(X_test)
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred2))
```
