Problem statement

In this dataset, you are presented pairs of phrases (an anchor and a target phrase) and asked to rate how similar they are on a scale from 0 (not at all similar) to 1 (identical in meaning).

Define target

```python
target = train.score
```

Combine train and test

```python
combi = train.drop(['score'], axis=1).append(test)
combi
```

Define X and y

```python
y = target
X = combi[: len(train)]
X_test = combi[len(train) :]
```

```python
import spacy_sentence_bert
nlp = spacy_sentence_bert.load_model('en_nli_roberta_large')

simularity = []

for i in range(len(X_test)):
    anchor = nlp(X_test['anchor'][i])
    text_target = nlp(X_test['target'][i])
    
    simularity.append(anchor.similarity(text_target))
    
print(len(simularity))
print(simularity)
```

Prepare submission

```python
submission['score'] = simularity
submission.to_csv("submission.csv", index=False)
submission
```
