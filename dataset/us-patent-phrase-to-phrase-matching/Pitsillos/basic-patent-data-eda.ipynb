# Patent EDA

```python
import pandas as pd
import plotly.express as px

train_df = pd.read_csv("../input/us-patent-phrase-to-phrase-matching/train.csv")
test_df = pd.read_csv("../input/us-patent-phrase-to-phrase-matching/test.csv")
```

# Basic Dataframe Information

```python
print(f"Train Dataframe shape: {train_df.shape}")
print(f"Test Dataframe shape: {test_df.shape}")

print("Train Dataframe Information\n---------------------------------------")
train_df.info()
```

```python
print("Null information on train dataframe")
print(train_df.isnull().sum())

num_unique_anchors = len(train_df["anchor"].unique())
num_unique_targets = len(train_df["target"].unique())
num_unique_contexts = len(train_df["context"].unique())

print(f"Number of unique anchors: {num_unique_anchors}")
print(f"Number of unique targets: {num_unique_targets}")
print(f"Number of unique contexts: {num_unique_contexts}")
```

# Histogram Plots

```python
px.histogram(train_df, x="context", title="Amount of samples in each context")
```

```python
px.histogram(train_df, x="context", color="anchor", title="Amount of samples in each context with anchor distribution")
```

```python
fig = px.histogram(train_df, x="score", nbins=len(train_df["score"].unique()), title="Amount of samples in each similarity score")
fig.update_xaxes(type='category', categoryorder="category ascending")
fig.show()
```

```python
train_group_context_unique_anchors = train_df.groupby(["context"]).nunique().reset_index()
fig = px.histogram(train_group_context_unique_anchors, x="context", y="anchor", title="Number of unique anchors per context")
fig.show()
fig = px.histogram(train_group_context_unique_anchors, x="context", y="target", title="Number of unique targets per context")
fig.show()
```
