```python
from fuzzywuzzy import fuzz
import pandas as pd
```

```python
train_df_raw = pd.read_csv('/kaggle/input/us-patent-phrase-to-phrase-matching/train.csv')
train_df_raw.head()
```

```python
train_df_raw['score'].value_counts()
```

```python
def x_round(x):
    return (round(x*4)/4)
def get_fuzzy_score(train_df_raw):
    train_df_raw['fuzzy_score'] = train_df_raw[['anchor', 'target']].apply(lambda x: fuzz.ratio(x['anchor'], x['target'])/100, axis=1)
    fuzz_result = train_df_raw['fuzzy_score'].to_list()
    fuzz_result = [x_round(f) for f in fuzz_result]
    train_df_raw['fuzzy_score'] = fuzz_result
    return train_df_raw

train_df_1 = get_fuzzy_score(train_df_raw)
```

```python
train_df_1['match'] = abs(train_df_1['fuzzy_score'] - train_df_1['score'])
accuracy = len(train_df_1[train_df_1['match']==0])/len(train_df_1)
print(f'With use of fuzzywuzzy, we are able to get {round(accuracy*100, 2)}')
```

```python
test_df_1 = get_fuzzy_score(test_df_raw)
```

```python
test_df_1.rename(columns = {'fuzzy_score': 'score'}, inplace=True)
```
