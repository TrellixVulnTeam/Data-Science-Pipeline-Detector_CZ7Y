# Library

```python
from sentence_transformers import SentenceTransformer, CrossEncoder
```

# Save Model and Pretraind

 embedding models list
 
 https://www.sbert.net/docs/pretrained_models.html#sentence-embedding-models/

```python
list_of_sentence_transformers = ['sentence-t5-xl', 'all-roberta-large-v1', 'all-mpnet-base-v2', 'all-distilroberta-v1', 'all-MiniLM-L12-v2']
```

```python
for model_name in list_of_sentence_transformers:
    model = SentenceTransformer(model_name)
    model.save(model_name)
```

Cross encoder list

https://www.sbert.net/docs/pretrained_cross-encoders.html

```python
list_of_cross_encoders = ['cross-encoder/stsb-roberta-base', 'cross-encoder/stsb-roberta-large', 'cross-encoder/stsb-distilroberta-base', 
                          'cross-encoder/ms-marco-MiniLM-L-12-v2', 'cross-encoder/quora-roberta-base', 'cross-encoder/quora-roberta-large',
                         'cross-encoder/nli-deberta-v3-base']
```

```python
for model_name in list_of_cross_encoders:
    model = CrossEncoder(model_name)
    model.save(model_name)
```
