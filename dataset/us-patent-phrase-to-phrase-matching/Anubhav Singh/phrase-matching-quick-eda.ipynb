```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from PIL import Image
```

```python
train_df=pd.read_csv('../input/us-patent-phrase-to-phrase-matching/train.csv')
test_df=pd.read_csv('../input/us-patent-phrase-to-phrase-matching/test.csv')
```

```python
print('Training Data Shape: ',train_df.shape)
print('Testing Data shape: ',test_df.shape)
```

```python
train_df.head()
```

```python
#Create Training Corpus
anchor_training_corpus=' '.join(train_df['anchor'])
target_training_corpus=' '.join(train_df['target'])
```

```python
#Create Testing Corpus
anchor_test_corpus=' '.join(test_df['anchor'])
target_test_corpus=' '.join(test_df['target'])
```

**Training Anchor Wordcloud**

```python
wordcloud = WordCloud(max_font_size=50, max_words=500, 
                      background_color="white").generate(anchor_training_corpus)
plt.figure(figsize=(10,10))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

**Training Target Wordcloud**

```python
wordcloud = WordCloud(max_font_size=50, max_words=500, 
                      background_color="white").generate(target_training_corpus)
plt.figure(figsize=(10,10))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

**Testing Anchor Wordcloud**

```python
wordcloud = WordCloud(max_font_size=50, max_words=500, 
                      background_color="white").generate(anchor_test_corpus)
plt.figure(figsize=(10,10))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

**Testing Target Wordcloud**

```python
wordcloud = WordCloud(max_font_size=50, max_words=500, 
                      background_color="white").generate(target_test_corpus)
plt.figure(figsize=(10,10))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
