# Imports

```python
from sklearn.model_selection import StratifiedKFold
```

# Cross validation folds

```python
skf = StratifiedKFold(n_splits = 10, shuffle= True, random_state=42)

df = pd.read_csv('/kaggle/input/us-patent-phrase-to-phrase-matching/train.csv')

df['kfold'] = -1

df['context_score'] = df['context']+"_"+df['score'].astype(str)

for fold_,(train_, valid_) in enumerate(skf.split(df, df.context_score)):
    df.loc[valid_,'kfold'] = fold_
    
#Drop context_score its not needed.
df.drop(columns=['context_score'], inplace = True)

```
