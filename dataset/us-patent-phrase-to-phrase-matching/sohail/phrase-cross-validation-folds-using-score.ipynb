# Kfold cross validation using score

```python
def create_folds(num_splits):
    
    skf = StratifiedKFold(n_splits = num_splits, shuffle = True, random_state=42)
    
    df = pd.read_csv('/kaggle/input/us-patent-phrase-to-phrase-matching/train.csv')
    
    df['kfold'] = -1
    
    for fold_, (train_, validation_) in enumerate(skf.split(df, df.score.astype(str))):
        df.loc[validation_, 'kfold'] = fold_
        
    return df
```

```python
df_train = create_folds(num_splits = 10)
df_train.head()
```

```python
#Submission
df_train.to_csv('phrase_cv_folds.csv', index = False)
```
