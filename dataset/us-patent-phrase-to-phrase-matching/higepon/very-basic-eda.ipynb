```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing
from pathlib import Path
        
data_dir = Path('/kaggle/input/us-patent-phrase-to-phrase-matching')
```

```python
train_df = pd.read_csv(data_dir / 'train.csv')
test_df = pd.read_csv(data_dir / 'test.csv')
train_df.head()
```

```python
num_train = len(train_df)
num_uniq_anchor = len(train_df.anchor.unique())
num_uniq_context = len(train_df.context.unique())
print(f'num train={num_train} num_uniq_anchor={num_uniq_anchor} num_uniq_context={num_uniq_context}')

num_test = len(test_df)
num_test_uniq_anchor = len(test_df.anchor.unique())
num_test_uniq_context = len(test_df.context.unique())
print(f'num test={num_test} num_uniq_anchor={num_test_uniq_anchor} num_uniq_context={num_test_uniq_context}')
```

### Context Distribution

```python
# Train
plt.figure(figsize=(40, 6))
_ = sns.countplot(x=train_df['context'])
```

```python
# Test
plt.figure(figsize=(20, 6))
_ = sns.countplot(x=test_df['context'])
```

### Context between train and test

```python
train_context = train_df.context.unique()
test_context = test_df.context.unique()

# all context from test appear in train.
set(test_context) <= set(train_context)
```

### Target length distribution

```python
_ = sns.countplot(x=train_df.target.apply(lambda s: len(s.split())))
```

### Score distribution

```python
_ = sns.countplot(x=train_df.score)
```
