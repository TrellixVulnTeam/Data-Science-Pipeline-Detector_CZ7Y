```python
import wandb, os

from kaggle_secrets import UserSecretsClient
user_secrets = UserSecretsClient()
api_key = user_secrets.get_secret("wandb")
wandb.login(key=api_key)

    
def download_artifacts_from_wandb(experiment : str):
    api = wandb.Api()
    
    try:
        artifact = api.artifact(f"gladiator/USPPPM-Kaggle/{experiment}:latest", type="model")
    except:
        artifact = api.artifact(f"gladiator/USPPPM-Kaggle/{experiment}:v0", type="model")
    artifact_dir = artifact.download(f"/kaggle/working/{experiment}")
```

```python
EXPERIMENTS = [
    '204_microsoft-cocolm-large_mse-msd',
    '205_microsoft-cocolm-large_pearson-msd'
]
```

```python
import shutil
for exp in EXPERIMENTS:
    download_artifacts_from_wandb(exp)
    for f in range(5):
        !cp ../input/cocolmlargevocab/dict.txt /kaggle/working/{exp}/fold_{f}/dict.txt
#         shutil.copyfile("../input/cocolmlargevocab/dict.txt", f"/kaggle/working/{exp}/fold_{f}/dict.txt")
        shutil.copyfile("../input/cocolmlargevocab/sp.model", f"/kaggle/working/{exp}/fold_{f}/sp.model")
```
