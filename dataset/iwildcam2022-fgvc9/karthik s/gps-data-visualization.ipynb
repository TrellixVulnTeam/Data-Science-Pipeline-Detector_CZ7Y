```python
import numpy as np
import pandas as pd
#!pip install folium
import folium
from branca.element import Figure
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import seaborn as sns; sns.set()
import csv
```

```python
gps = '../input/iwildcam2022-fgvc9/metadata/metadata/gps_locations.json'
gps_pd = pd.read_json(gps)
data = gps_pd.transpose()
```

```python
ps_pd = data[['latitude','longitude']]
locationlist = ps_pd.values.tolist()
ps_pd.head()
```

# Clustering for better Visualizing 
Since I was unable to visualize all the GPS coordinates with a single GPS anchor I have clustered the GPS coordinates.

```python
kmeans = KMeans(n_clusters = 6, init ='k-means++')
kmeans.fit(ps_pd[ps_pd.columns[0:2]]) # Compute k-means clustering.
ps_pd['cluster_label'] = kmeans.fit_predict(ps_pd[ps_pd.columns[0:2]])
centers = kmeans.cluster_centers_ # Coordinates of cluster centers.
labels = kmeans.predict(ps_pd[ps_pd.columns[0:2]]) # Labels of each point
ps_pd.head(5)
```

# Visualizing the MAP using Folium Marker

```python
map = folium.Map(width=1000, height=500,location=[ 0.38959009,  36.87179589], zoom_start=11)
for point in range(0, len(locationlist)):
    folium.Marker(locationlist[point],popup=folium.Popup(max_width=450).add_child(
        folium.Vega(data, width=450, height=250)
    ),icon=folium.Icon(color="green")).add_to(map)
map
```
