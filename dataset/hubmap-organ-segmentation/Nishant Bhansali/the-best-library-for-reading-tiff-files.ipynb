# Analysis of time taken by different packages to read a .tiff file

this include reading the file, changing the channel order if necessary and then finally converting it into a pytorch tensor

# Results
- skimage :17.06 ms
- tifffile : 17.03 ms
- PIL : 63.5ms
- Rasterio: 234 ms
- Opencv: 111ms

```python
import numpy as np
import pandas as pd
import cv2
import torch 
import matplotlib.pyplot as plt
from skimage import io,img_as_float
from torchvision.transforms import ToTensor
import rasterio
from rasterio.plot import show
from PIL import Image


PATH = "/kaggle/input/hubmap-organ-segmentation/train_images/28189.tiff"
```

## SKIMAGE -> torch Tensor
17.6 ms

```python
%%timeit

img = io.imread(PATH)
img = torch.tensor(img)
```

## Rasterio -> tensor

```python
%%timeit
with rasterio.open(PATH) as image:
    image_array = image.read()
    
torch_image = ToTensor()(image_array)
```

## Rasterio
2.04ms
31.03ms
38.07ms

```python
%%timeit
s = rasterio.open(PATH)
img = s.read()
img = torch.tensor(img)
```

## PIL Image
- 1.22ms only reading using image.open
- 63.5ms to convert it to first numpy array and then to torch tensor


```python
%%timeit
pil_image = Image.open(PATH)
img = torch.tensor(np.array(pil_image))
```
