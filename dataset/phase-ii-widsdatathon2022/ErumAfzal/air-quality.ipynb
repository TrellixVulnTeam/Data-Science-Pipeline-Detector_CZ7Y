```python
df_19 = df_19.loc[:, df_19.isnull().mean() < .8]
df_19.head()
```

```python
cat=[]
num=[]
for i in df_19.columns:
    if df_19[i].dtype=='object':
        cat.append(i)
    elif df_19[i].dtype=='float64':
        num.append(i)
        
```

```python
df_19[cat].info()
```

```python
df_19[num].info()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
df_19[num] = imputer.fit_transform(df_19[num])
df_19[num]
```
