<h1 style="color:#189AB4"><strong>The First Question!</strong> What is the Null Distribution Column Wise?</h1>

```python
df1 = pd.DataFrame(df.isna().sum().sort_values(ascending=False))
df1['null']=df1.index
df1['count']=df1.iloc[:,:-1]
df1.reset_index(drop=True, inplace=True)
df1 = df1.drop(df1.columns[[0]],axis = 1)
plt.title('Null Distribution Column-Wise')
ax = sns.barplot(y='null',x='count',data=df1.head(14))
for i in ax.containers:
    ax.bar_label(i)
```

Out of 129470 rows 128811 values in column **LEAD_UG_PER_CUBIC_METER** & 126163 values in column **BENZENE_PPBC** are missing
- It is better to drop these columns as they contain mostly null values
