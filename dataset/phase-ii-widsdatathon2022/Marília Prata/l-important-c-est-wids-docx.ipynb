```python
#Code by Stack Overflow https://stackoverflow.com/questions/29309085/read-docx-files-via-python¶
#Answered by Zaid.mohammed in Stack Overflow Nov 30 '19 at 12:43

def main():
    try:
        doc = docx.Document('../input/phase-ii-widsdatathon2022/critical_health/critical_health/Health-Related Datasets for WiDS Phase 2 Datathon.docx')  # Creating word reader object.
        data = ""
        fullText = []
        for para in doc.paragraphs:
            fullText.append(para.text)
            data = '\n'.join(fullText)

        print(data)

    except IOError:
        print('There was an error opening the file!')
        return


if __name__ == '__main__':
    main()
```
