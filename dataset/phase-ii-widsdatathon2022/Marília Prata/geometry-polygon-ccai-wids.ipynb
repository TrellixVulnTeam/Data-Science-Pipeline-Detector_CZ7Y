```python
df = pd.read_csv("/kaggle/input/phase-ii-widsdatathon2022/ccai/ccai/data.csv", delimiter=',', encoding='utf8')
df.tail(2)
```

```python
#Code by Alexander Bader  https://www.kaggle.com/code/alexanderbader/use-geo-data

import geopandas as gpd
from shapely.geometry import Point

#ds=pd.read_csv("../input/forbes-billionaires-2021-30/forbes_billionaires_geo.csv")

from shapely import wkt

df['geometry'] = df['geometry'].apply(wkt.loads)

world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
df = gpd.GeoDataFrame(df).set_geometry('geometry')

base = world.plot(color='white', edgecolor='black', figsize=(16, 9))

df.plot(ax=base, marker='o', color='magenta', markersize=1);
```

```python
#Code by Alexander Bader  https://www.kaggle.com/code/alexanderbader/use-geo-data

resid = df.groupby("city_name").count().address[(df.groupby("city_name").count().address>1).values].sort_values()
```

```python
#Code by Alexander Bader  https://www.kaggle.com/code/alexanderbader/use-geo-data

res_plot = df[df.city_name.isin(resid.index)]
res_plot = gpd.GeoDataFrame(res_plot.groupby("city_name").count().address)
```

```python
#Code by Alexander Bader  https://www.kaggle.com/code/alexanderbader/use-geo-data

cc=[]
for i in res_plot.index:
    cc.append(df[df.city_name==i].reset_index().geometry[0])
    
res_plot["geometry"]=cc
res_plot.set_geometry('geometry')
```

```python
#Code by Alexander Bader  https://www.kaggle.com/code/alexanderbader/use-geo-data

base = world.plot(color='white', edgecolor='gray', figsize=(35, 15))

ax = res_plot.plot(ax=base, marker='o', cmap='Pastel1', markersize=res_plot.address*20);
```
