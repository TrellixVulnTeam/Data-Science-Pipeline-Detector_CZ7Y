```python
f = open("/kaggle/input/expedia-hotel-recommendations/train.csv", "r")
a = f.readline().strip().split(',')
print(a[16])
print(a[-1])
print(a[-6])
```

```python
from collections import defaultdict

d = defaultdict(lambda: defaultdict(int))
n = 0
while(1):
    line = f.readline().strip()
    n += 1
    
    if n % 1000000 == 0:
        print('Read {} lines...'.format(n))
    
    if line == '':
        break
        
    arr = line.split(',')
    destination = arr[16]
    score = 20 if arr[-6] == 1 else 1
    hotel = arr[-1]
    if destination != '':
        d[destination][hotel] += score
```

```python
f.close()
f2 = open("/kaggle/input/expedia-hotel-recommendations/test.csv","r")
a = f2.readline().split(',')
print(a[0])
print(a[17])
```

```python
out = open("submission.csv","w")
out.write("id,hotel_cluster\n")

from heapq import nlargest
from operator import itemgetter

n = 0
while(1):
    line = f2.readline().strip()
    n += 1
    
    if n % 1000000 == 0:
        print('Read {} lines...'.format(n))
    
    if line == '':
        break
    
    arr = line.split(',')
    idx = arr[0]
    hotels = d[arr[17]]
    topitems = nlargest(5, hotels.items(), key=itemgetter(1))
    out.write(str(idx) + ',')
    for i in range(len(topitems)):
        out.write(' ' + topitems[i][0])
    out.write('\n')

out.close()
```
