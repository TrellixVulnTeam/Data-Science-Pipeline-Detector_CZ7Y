On transforme le type de toute les colonne "object" en "category"

```python
for col_name in df.columns:
    if(df[col_name].dtype == 'object'):
        df[col_name]= df[col_name].astype('category')
        df[col_name] = df[col_name].cat.codes

display(df)
```

```python
X = df.drop(["Transported"], axis=1)
y = df.Transported
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
clf = LogisticRegression(random_state=0, max_iter=10000).fit(X_train, y_train)
ypred = clf.predict(X_test)
cm = confusion_matrix(y_test, ypred, labels=clf.classes_)
ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=clf.classes_).plot()
```

```python
print('Score de la regression logistique: {:.2f}'.format(accuracy_score(y_test, clf.predict(X_test))))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, ypred))
```
