```python
train = pd.read_csv('../input/spaceship-titanic/train.csv')
print("train -DONE!")
test = pd.read_csv('../input/spaceship-titanic/test.csv')
print("test -DONE!")
```

```python
import seaborn as sns
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib import style
```

```python
grid = sns.FacetGrid(train, col='Transported', row='VIP', size=2.2, aspect=1.6)
grid.map(plt.hist, 'Age', alpha=.5, bins=20)
grid.add_legend();
```

```python
features = ["HomePlanet", "CryoSleep",  "Destination",'Age','RoomService', 'FoodCourt', 'ShoppingMall','Spa']
X_train = pd.get_dummies(train[features])
Y_train = train["Transported"]
X_test = pd.get_dummies(test[features])

X_train[ X_train.isnull() == True] = 0
X_test [X_test.isnull() == True] = 0
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_train, Y_train)

Y_prediction = random_forest.predict(X_test)

random_forest.score(X_train, Y_train)
acc_random_forest = round(random_forest.score(X_train, Y_train) * 100, 2)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train, Y_train)

Y_pred = logreg.predict(X_test)

acc_log = round(logreg.score(X_train, Y_train) * 100, 2)
```

```python
from sklearn.linear_model import SGDClassifier
from sklearn import linear_model
sgd = linear_model.SGDClassifier(max_iter=5, tol=None)
sgd.fit(X_train, Y_train)
Y_pred = sgd.predict(X_test)

sgd.score(X_train, Y_train)

acc_sgd = round(sgd.score(X_train, Y_train) * 100, 2)
```

```python
from sklearn.tree import DecisionTreeClassifier
decision_tree = DecisionTreeClassifier() 
decision_tree.fit(X_train, Y_train)  
Y_pred = decision_tree.predict(X_test)  
acc_decision_tree = round(decision_tree.score(X_train, Y_train) * 100, 2)
```

```python
results = pd.DataFrame({
    'Model': ['Logistic Regression', 
              'Random Forest',
              'Stochastic Gradient Decent', 
              'Decision Tree'],
    'Score': [ acc_log, 
              acc_random_forest, 
              acc_sgd, acc_decision_tree]})
result_df = results.sort_values(by='Score', ascending=False)
result_df = result_df.set_index('Score')
result_df.head(9)
```
