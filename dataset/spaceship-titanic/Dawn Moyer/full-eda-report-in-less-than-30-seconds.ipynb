```python
import pandas as pd
!pip install dataprep
from dataprep.eda import create_report
```

### Read the train file

```python
df = pd.read_csv('../input/spaceship-titanic/test.csv')
print(df.shape)
```

### Run the report creation

```python
report = create_report(df, title='Spaceship Titanic Training Data')
```

### You have a choice of three outputs
 1. Write to a file - html is great
 2. Open the report in a browser tab (outside of the kaggle environment)
 3. Display the report inline


```python
report.save('../Spaceship_Titanic_EDA.html')

report.show_browser()

report.show()
```
