```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from keras import models,layers
```

```python
train = pd.read_csv('/kaggle/input/spaceship-titanic/train.csv')
test = pd.read_csv('/kaggle/input/spaceship-titanic/test.csv')
```

```python
Y_train = train["Transported"]
```

```python
train['PassengerGroup'] = train['PassengerId'].apply(lambda x: x.split('_')[1])
test['PassengerGroup'] = test['PassengerId'].apply(lambda x: x.split('_')[1])

train['HomePlanet'].fillna(train.HomePlanet.mode()[0], inplace=True) #Earth
test['HomePlanet'].fillna(test.HomePlanet.mode()[0], inplace=True) #Earth

train['CryoSleep'].fillna(False, inplace=True)
test['CryoSleep'].fillna(False, inplace=True)

train['VIP'].fillna(train.VIP.mode()[0], inplace=True) 
test['VIP'].fillna(test.VIP.mode()[0], inplace=True)

train['Destination'].fillna(train.Destination.mode()[0], inplace=True)
test['Destination'].fillna(test.Destination.mode()[0], inplace=True)

train['Age'].fillna(train.Age.mean(), inplace=True)
train[['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']] =\
train[['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']].fillna(0)

test['Age'].fillna(test.Age.mean(), inplace=True)
test[['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']] =\
test[['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']].fillna(0)
```

```python
valores = ["HomePlanet","CryoSleep","Destination","VIP","Spa", "VRDeck","Age","RoomService","FoodCourt","ShoppingMall","PassengerGroup"]

X_train = pd.get_dummies(train[valores])
X_test = pd.get_dummies(test[valores])

X_train
```

RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier

def rf(X_train, Y_train, n_estimators=400):
    clf = RandomForestClassifier(n_estimators=n_estimators, n_jobs=-1,random_state=1)
    clf.fit(X_train, Y_train)
    return clf

random_forest_en = rf(X_train, Y_train , n_estimators=400)
```

```python
def print_result(model, x_train, y_train):

    print('Acierto en el entrenamiento ', model.score(x_train, y_train))

print_result(random_forest_en, X_train, Y_train)
```

```python
predrf = random_forest_en.predict(X_test)
print('Resultado: ',predrf)
```

XGBOOST

```python
from xgboost import XGBClassifier
xg = XGBClassifier(learning_rate=0.001, n_estimators=200,
                   max_depth= 10, min_child_weight= 1, 
                   colsample_bytree= 0.6, gamma= 0.0, 
                   reg_alpha= 0.001, subsample= 0.8
                  )
xg.fit(X_train, Y_train)
xg_predictions = xg.predict(X_test)
scores = xg.score(X_train, Y_train)

print(scores)
```
