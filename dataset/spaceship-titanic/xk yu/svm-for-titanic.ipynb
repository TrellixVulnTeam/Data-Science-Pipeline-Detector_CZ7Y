```python
import pandas as pd
from sklearn.svm import SVC
import numpy as np
import joblib

# 数据读取及处理
data = pd.read_csv('../input/spaceship-titanic/train.csv',index_col=0)
#print(data)

data.fillna({'CryoSleep':False,'VIP':False,'Age':int(data['Age'].mean()),'RoomService':0.0,'FoodCourt':0.0,\
            'ShoppingMall':0.0,'Spa':0.0,'VRDeck':0.0},inplace=True)

X = data.iloc[:,[1,4,5,6,7,8,9,10]]
y = data.iloc[:,-1]



#构建模型
svm_titan = SVC()
svm_titan.fit(X,y)
joblib.dump(svm_titan,'./svm_tatan.pkl')
```

```python
svm_module = joblib.load('./svm_tatan.pkl')
test_data = pd.read_csv('../input/spaceship-titanic/test.csv')
print(test_data)
test_data.fillna({'CryoSleep':False,'VIP':False,'Age':int(test_data['Age'].mean()),'RoomService':0.0,'FoodCourt':0.0,\
            'ShoppingMall':0.0,'Spa':0.0,'VRDeck':0.0},inplace=True)
#result = test_data.iloc[:,0]
result = pd.DataFrame(test_data.iloc[:,0])
temp = svm_module.predict(test_data.iloc[:,[2,5,6,7,8,9,10,11]])
result.insert(1,'Transported',temp)
result.to_csv('./submission.csv',index=False)
print(result)
#result['Transported'] = test_data.apply(dect,axis=1)
#print(result)
```
