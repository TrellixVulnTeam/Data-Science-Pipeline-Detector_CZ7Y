```python
from pycaret.classification import *
from pathlib import Path
```

```python
path = Path('../input/spaceship-titanic')
train_df = pd.read_csv(path/'train.csv')
test_df = pd.read_csv(path/'test.csv')
```

```python
train_df.head()
```

```python
clf = setup(data = train_df, target = 'Transported', use_gpu=True)
```

```python
cat_boost_model = create_model('catboost')
```

```python
ensembled_cat_boost_model = ensemble_model(cat_boost_model, method='Bagging', n_estimators = 25, 
                                           return_train_score=True)
#tuned_cat_boost_model = tune_model(cat_boost_model)
```

```python
preds = predict_model(ensembled_cat_boost_model, test_df)
```

```python
preds.shape
```

```python
preds.head()
```

```python
sub_df = preds[['PassengerId', 'Label']]
sub_df.columns = ['PassengerId', 'Transported']
```
