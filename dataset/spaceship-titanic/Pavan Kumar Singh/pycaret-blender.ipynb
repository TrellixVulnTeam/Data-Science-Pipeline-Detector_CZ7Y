```python
from pycaret.classification import *
from pathlib import Path
```

```python
path = Path('../input/spaceship-titanic')
train_df = pd.read_csv(path/'train.csv')
test_df = pd.read_csv(path/'test.csv')
```

```python
train_df.head()
```

```python
clf = setup(data = train_df, target = 'Transported', use_gpu=True)
```

```python
#top_3_models = compare_models(n_select=3)
```

```python
light_gradient_boost_model = create_model('lightgbm')
catboost_model = create_model('catboost')
gradient_boost_model = create_model('gbc')
```

```python
#top_3_models
```

```python
#tuned_top_3_models = [tune_model(i) for i in top_3_models]
```

```python
blended_model = blend_models([light_gradient_boost_model, catboost_model, gradient_boost_model])
```

```python
stacked_model = stack_models([light_gradient_boost_model, catboost_model, gradient_boost_model])
```

```python
preds = predict_model(stacked_model, test_df)
```

```python
sub_df = preds[['PassengerId', 'Label']]
sub_df.columns = ['PassengerId', 'Transported']
```
