```python
#ライブラリをインポート
import pandas as pd
```

kaggle/spaceship-titanicというフォルダをローカルに作って、そこにtrain.csvとtest.csvを置いておきます。

```python
#CSVファイル読み込み。
df_train = pd.read_csv('kaggle/spaceship-titanic/train.csv')
#読み込んだデータの列名一覧を表示。
df_train.columns
```

```python
#欠損値が多い列を探す。
missing_data = df_train.isnull().sum().sort_values(ascending=False)
missing_data = pd.concat([missing_data], axis=1, keys=['missing_data'])
missing_data.head(10)
```

```python
#データの大きさを確認する。
df_train.shape
```

```python
#VIPとAgeとTransportedの３つのいずれかが欠損している行は除く。
df_train = df_train.dropna(subset=["VIP","Age","Transported"])
#データの大きさを再度確認する。
df_train.shape
```

```python
# X_trainにdf_trainのVIPとAgeだけ代入。
X_train = df_train.loc[:,["VIP","Age"]]

# Y_trainにdf_trainのTransportedを代入。
Y_train = df_train["Transported"]
```

```python
from sklearn.linear_model import LogisticRegression

#訓練データX_trainとY_trainを使いロジスティック回帰による分類モデルを作成。
lg = LogisticRegression()
lg.fit(X_train, Y_train)
```

```python
#テストデータ読み込み。
X_test = pd.read_csv('kaggle/spaceship-titanic/test.csv')
Id = X_test["PassengerId"]
#テストデータからも、訓練データと同じ列を抽出。
X_test = X_test.loc[:,["VIP","Age"]].fillna(X_test.mean())
```

```python
#モデルを使って予測値を出す。
Y_pred = lg.predict(X_test)
```

```python
#予測値をcsvファイルとして出力。
Predict = pd.DataFrame({
        'PassengerId': Id,
        "Transported": Y_pred
    })

Predict.to_csv('kaggle/spaceship-titanic/el.predict.csv', index=False, encoding="SHIFT-JISx0213")
```
