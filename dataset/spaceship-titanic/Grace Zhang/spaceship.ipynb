```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import math
```

```python
def change_text(column, df, wanted=False, started_from=1): # Function to change text column to numbers
    the_column = df[column]
    u = {}
    num = started_from
    for unique in set(the_column):
        if type(unique) == float:
            u[unique] = float("nan")
        else:
            u[unique] = num
            num += 1
#             num += 1
#     for unique in set(the_column):
    
    df[column] = [u[x] for x in df[column]]
    print(set(the_column))
    if wanted:
        print(u)
```

```python
def change_back(the_dic, df): # Function to change it back
    new_dic = {}
    for key, value in the_dic.items():
        new_dic[value] = key
    df["Transported"] = [new_dic[x] for x in df["Transported"]]
```

# **Get ready for data** ðŸŽ†

```python
df = pd.read_csv("../input/spaceship-titanic/train.csv") # Main dataframe
df.pop("Name")
# y = pd.DataFrame(X.pop("Transported"))
# X.head()

X, tX = train_test_split(df, test_size=0.3) # Split train X and test X
y = pd.DataFrame(X.pop("Transported")) # Pop the train y
ty = pd.DataFrame(tX.pop("Transported")) # Pop the test y

test_X = pd.read_csv("../input/spaceship-titanic/test.csv") #Import submission X
test_X.pop("Name") # Remove name submission X
```

```python
# texts = ["HomePlanet", "CryoSleep", "Cabin", "Destination", "VIP"] # Get columns ready
texts = ["HomePlanet", "CryoSleep", "Destination", "VIP"] # Get columns ready
```

```python
X.pop("Cabin")
tX.pop("Cabin")
test_X.pop("Cabin")

X.pop("PassengerId")
tX.pop("PassengerId")
test_X.pop("PassengerId")
```

```python
X = pd.get_dummies(X)
tX = pd.get_dummies(tX)
test_X = pd.get_dummies(test_X)
```

```python
col = list(X.columns)
for c in col:
    try:
        X[c] = X[c].fillna(X[c].mean()) # Fill the null/nan objects of train X
        tX[c] = tX[c].fillna(tX[c].mean()) # Fill the null/nan objects of test X
        test_X[c] = test_X[c].fillna(test_X[c].mean())
    except:
        pass
```

# **Training**ðŸ‘“

### **KNeighborsClassifier**

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
time = []
predict = []
for num in range(1, 50):
    knc = KNeighborsClassifier(n_neighbors=num)
    knc.fit(X, y.values.ravel())
    time.append(knc.score(tX, ty.values.ravel()))
    predict.append(knc.predict(test_X))
#     print(num, rfc.score(tX, ty.values.ravel()))
print(time)
print(time[np.argmax(time)])

print(predict[np.argmax(time)])
mainone_prediction = predict[np.argmax(time)]
```

### **RandomForestClassifier**

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
time = []
predict = []
for num in range(1, 50):
    rfc = RandomForestClassifier(n_estimators=num, random_state=0)
    rfc.fit(X, y.values.ravel())
    time.append(rfc.score(tX, ty.values.ravel()))
    predict.append(rfc.predict(test_X))
#     print(num, rfc.score(tX, ty.values.ravel()))
print(time)
print(time[np.argmax(time)])

print(predict[np.argmax(time)])
maintwo_prediction = predict[np.argmax(time)]
```

### **DecisionTreeClassifier**

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
time = []
predict = []
for num in range(1, 50):
    dtc = DecisionTreeClassifier(max_depth=num)
    dtc.fit(X, y.values.ravel())
    time.append(dtc.score(tX, ty.values.ravel()))
    predict.append(dtc.predict(test_X))
#     print(num, rfc.score(tX, ty.values.ravel()))
print(time)
print(time[np.argmax(time)])

print(predict[np.argmax(time)])
mainthree_prediction = predict[np.argmax(time)]
```

### **AdaBoostClassifier**

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
time = []
predict = []
for num in range(1, 50):
    abc = AdaBoostClassifier(n_estimators=num, random_state=1)
    abc.fit(X, y.values.ravel())
    time.append(abc.score(tX, ty.values.ravel()))
    predict.append(abc.predict(test_X))
#     print(num, rfc.score(tX, ty.values.ravel()))
print(time)
print(time[np.argmax(time)])

print(predict[np.argmax(time)])
mainfour_prediction = predict[np.argmax(time)]
```
