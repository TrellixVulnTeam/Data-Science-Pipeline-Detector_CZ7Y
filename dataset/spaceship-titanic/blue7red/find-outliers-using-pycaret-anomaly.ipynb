```python
train = pd.read_csv(TRAIN_PATH)
```

```python
def getOutliersBasic(df):
    setup(data=df,silent=True)
    knn = create_model('knn')
    knn_df = assign_model(knn)
    abnormalData = knn_df[knn_df.Anomaly == 1]
    return abnormalData
```

```python
outliers = getOutliersBasic(train)
print("len =",len(outliers))
outliers[:5]
```
