```python
import numpy as np
import pandas as pd 

TRAIN_PATH = "../input/spaceship-titanic/train.csv"
TEST_PATH = "../input/spaceship-titanic/test.csv"
```

```python
def addSplitedColumn(df,COL,SPLIT_SIZE,SEP):
    for i in range(SPLIT_SIZE):
        df[COL + "_" + str(i)] = df[COL].str.split(pat=SEP,expand=True)[i]

    df = df.drop([COL],axis=1)
    
    return df
    
```

```python
train = pd.read_csv(TRAIN_PATH)
test = pd.read_csv(TEST_PATH)
```

```python
COL = "Cabin"
SPLIT_SIZE = 3
SEP = "/"
```

```python
train = addSplitedColumn(train,COL,SPLIT_SIZE,SEP)
train.head()
```

```python
test = addSplitedColumn(test,COL,SPLIT_SIZE,SEP)
test.head()
```
