# Before 

```python
from IPython.display import clear_output
!pip3 install -q -U git+https://github.com/mljar/mljar-supervised.git@master
clear_output()

import numpy as np
import pandas as pd

from supervised.automl import AutoML

TRAIN_PATH = "../input/spaceship-titanic/train.csv"
TEST_PATH = "../input/spaceship-titanic/test.csv"
SAMPLE_SUBMISSION_PATH = "../input/spaceship-titanic/sample_submission.csv"
SUBMISSION_PATH = "submission.csv"

ID = "PassengerId"
TARGET = "Transported"

MODEL_MODE = "Compete"
MODEL_METRIC = "accuracy"
MODEL_ALGORITHMS = ["CatBoost","Xgboost", "LightGBM"]
MODEL_TOTAL_TIME_LIMIT = 3*60
MODEL_TIME_LIMIT = 3*60
MODEL_START_RANDOM_MODELS = 3
MODEL_HILL_CLIMBING_STEPS = 3
MODEL_TOP_MODELS_TO_IMPROVE = 3
MODEL_EXPLAIN_LEVEL = 1
MODEL_NJOBS = 14
```

# Build

```python
train = pd.read_csv(TRAIN_PATH)

X = train.drop([ID,TARGET],axis=1)
y = train[TARGET]

automl = AutoML(
    mode= MODEL_MODE, 
    total_time_limit=MODEL_TOTAL_TIME_LIMIT,
    eval_metric=MODEL_METRIC,
    algorithms=MODEL_ALGORITHMS,
    model_time_limit=MODEL_TIME_LIMIT,
    start_random_models=MODEL_START_RANDOM_MODELS,
    hill_climbing_steps=MODEL_HILL_CLIMBING_STEPS,
    top_models_to_improve=MODEL_TOP_MODELS_TO_IMPROVE,
    golden_features=True,
    features_selection=True,
    stack_models=True,
    train_ensemble=True,
    explain_level=MODEL_EXPLAIN_LEVEL,
    n_jobs=MODEL_NJOBS,
    kmeans_features = True
)
automl.fit(X, y)
```

# After

```python
test = pd.read_csv(TEST_PATH)

X_test = test.drop([ID],axis=1)

preds = automl.predict(X_test)

sub = pd.read_csv(SAMPLE_SUBMISSION_PATH)
sub[TARGET] = preds.astype(bool)
sub.to_csv(SUBMISSION_PATH,index = False)
sub.head()
```
