```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import MinMaxScaler, LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from sklearn.impute import SimpleImputer
```

```python
data = pd.read_csv('../input/spaceship-titanic/train.csv')
```

```python
data.head()
```

```python
data.columns
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data = data.drop(['PassengerId', 'Name'], axis=1)
```

```python
Cabin_deck = []
for i in data['Cabin']:
    try:
        Cabin_deck.append(i[0])
    except:
        Cabin_deck.append(np.nan)
data['Cabin_deck'] = Cabin_deck

Cabin_side = []
for i in data['Cabin']:
    try:
        Cabin_side.append(i[-1])
    except:
        Cabin_side.append(np.nan)
data['Cabin_side'] = Cabin_side
```

```python
data = data.drop('Cabin', axis=1)
```

```python
numeric = ['Age','RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']
categorical = ['HomePlanet', 'CryoSleep', 'Destination', 'VIP', 'Cabin_deck', 'Cabin_side']
```

```python
for col in categorical:
    data[[col]] = SimpleImputer(strategy='most_frequent').fit_transform(data[[col]])

for col in numeric:
    data[[col]] = SimpleImputer(strategy='mean').fit_transform(data[[col]])
```

```python
data['Home to Dest'] = data['HomePlanet'] + ' to ' + data['Destination']
```

```python
categorical.append('Home to Dest')
```

```python
data.isnull().sum()
```

```python
data['Expenses'] = data['RoomService'] + data['FoodCourt'] + data['ShoppingMall'] + data['Spa'] + data['VRDeck']
```

```python
numeric.append('Expenses')
```

```python
data[['Expenses']].groupby(data['Transported']).mean().reset_index()
```

```python
plt.hist(data['Expenses'], bins=5, color='forestgreen', alpha=0.7)
plt.show()
```

```python
sns.barplot(data['Transported'], data['Expenses'])
plt.show()
```

```python
plt.hist(data['Age'], bins=5, color='forestgreen', alpha=0.7)
plt.show()
```

```python
age_cat = []
for i in data['Age']:
    if i>60:
        age_cat.append('Senior')
    elif i>18:
        age_cat.append('Adult')
    else:
        age_cat.append('Children and Teens')
data['Age Category'] = age_cat
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
sns.countplot(data=data, x='HomePlanet', hue='Transported')

plt.subplot(1,3,2)
sns.countplot(data=data, x='Destination', hue='Transported')

plt.subplot(1,3,3)
sns.countplot(data=data, x='Home to Dest', hue='Transported')
plt.xticks(rotation=90)

plt.show()
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.countplot(data=data, x='Cabin_deck', hue='Transported')

plt.subplot(1,2,2)
sns.countplot(data=data, x='Cabin_side', hue='Transported')

plt.show()
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.barplot(data['Transported'], data['Age'])

plt.subplot(1,2,2)
sns.countplot(data=data, x='Age Category', hue='Transported')

plt.show()
```

```python
for col in categorical:
    data[col] = LabelEncoder().fit_transform(data[col])
```

```python
for col in numeric:
    data[[col]] = StandardScaler().fit_transform(data[[col]])
```

```python
data.head()
```

```python
x = data.drop(['Transported', 'Home to Dest', 'Expenses', 'Age Category'], axis=1)
y = data[['Transported']]
```

```python
x.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0, test_size=0.2)
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
model = RandomForestClassifier().fit(x_train, y_train['Transported'])
```

```python
y_hat = model.predict(x_test)
```

```python
confusion_matrix(y_test['Transported'], y_hat)
```

```python
accuracy_score(y_test['Transported'], y_hat)
```

```python
print(classification_report(y_test['Transported'], y_hat))
```
