```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv('../input/spaceship-titanic/train.csv')
test = pd.read_csv('../input/spaceship-titanic/test.csv')
sample = pd.read_csv('../input/spaceship-titanic/sample_submission.csv')
```

```python
train.isna().sum()
```

```python
test.isna().sum()
```

```python
train.drop(['PassengerId','Name'],axis=1, inplace=True)
test.drop(['PassengerId','Name'],axis=1, inplace=True)
```

```python
display(train, test)
```

```python
y = train.pop('Transported')
```

```python
display(train, test)
```

```python
cat = [col for col in test.columns if test[col].nunique() < 15]
cont = [col for col in test.columns if not col in cat]
display(cat, cont)
```

```python
from sklearn.impute import SimpleImputer
Impute = SimpleImputer(strategy='most_frequent')
train[train.columns] = Impute.fit_transform(train)
train.isna().sum()
```

```python
test[test.columns] = Impute.transform(test)
test.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
for col in cat:
    le = LabelEncoder()
    train[col] = le.fit_transform(train[col])
    test[col] = le.transform(test[col])
```

```python
train.drop(['Cabin'], axis=1, inplace=True)
test.drop(['Cabin'], axis=1, inplace=True)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(train,y)
```

```python
ypred = lr.predict(test)
```

```python
sample['Transported'] = ypred
```

```python
sample.to_csv('submission.csv', index=False)
```
