```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

test_data = pd.read_csv("../input/spaceship-titanic/test.csv")
train_data = pd.read_csv("../input/spaceship-titanic/train.csv")

test_data2 = test_data
test_data.head()
```

```python
train_data = train_data.drop(['Name'], axis = 1)
train_data.head()
```

```python
#Check for na per column
nans = train_data.isna().sum().sort_values(ascending=False)
missing_data = pd.concat([nans], axis=1, keys=["Total"])
missing_data
```

```python
#Fill na with 'other'
train_data['CryoSleep'] = train_data['CryoSleep'].fillna('other')
train_data['VIP'] = train_data['VIP'].fillna('other')
train_data['HomePlanet'] = train_data['HomePlanet'].fillna('other')
train_data['Cabin'] = train_data['Cabin'].fillna('other')
train_data['Destination'] = train_data['Destination'].fillna('other')

#Average of age and monetary columns
imputer_cols = ["Age", "FoodCourt", "ShoppingMall", "Spa", "VRDeck" ,"RoomService"]
imputer = SimpleImputer(strategy="mean" )
imputer.fit(train_data[imputer_cols])
train_data[imputer_cols] = imputer.transform(train_data[imputer_cols])
```

```python
#Check for na per column
nans = train_data.isna().sum().sort_values(ascending=False)
missing_data = pd.concat([nans], axis=1, keys=["Total"])
missing_data
```

# Train data

```python
from sklearn.preprocessing import OrdinalEncoder

#HomePlanet_trf
HomePlanet_trf=['Europa', 'Earth', 'Mars', 'other']
HomePlanet=OrdinalEncoder(categories=[HomePlanet_trf])
HomePlanet.fit_transform(train_data[["HomePlanet"]])

train_data[["HomePlanet"]]=HomePlanet.fit_transform(train_data[["HomePlanet"]])

#CryoSleep
CryoSleep_trf=[False, True, 'other']
CryoSleep=OrdinalEncoder(categories=[CryoSleep_trf])
CryoSleep.fit_transform(train_data[["CryoSleep"]])

train_data[["CryoSleep"]]=CryoSleep.fit_transform(train_data[["CryoSleep"]])

#Destination
Destination_trf=['TRAPPIST-1e', 'PSO J318.5-22', '55 Cancri e', 'other']
Destination=OrdinalEncoder(categories=[Destination_trf])
Destination.fit_transform(train_data[["Destination"]])

train_data[["Destination"]]=Destination.fit_transform(train_data[["Destination"]])

#VIP
VIP_trf=[False, True, 'other']
VIP=OrdinalEncoder(categories=[VIP_trf])
VIP.fit_transform(train_data[["VIP"]])

train_data[["VIP"]]=CryoSleep.fit_transform(train_data[["VIP"]])

#Transported
Transported_trf=[False, True]
Transported=OrdinalEncoder(categories=[Transported_trf])
Transported.fit_transform(train_data[["Transported"]])

train_data[["Transported"]]=Transported.fit_transform(train_data[["Transported"]])
```

```python
train_data.head()
```

# **Convert Number columns values into range represented by numbers**

```python
train_data.loc[(train_data.Age < 20),  'Age_cat'] = 1
train_data.loc[(train_data.Age >= 20) & (train_data.Age < 40),  'Age_cat'] = 2
train_data.loc[(train_data.Age >= 40) & (train_data.Age < 60),  'Age_cat'] = 3
train_data.loc[(train_data.Age > 60), 'Age_cat'] = 4

train_data = train_data.drop('Age',axis=1)

train_data['Age_cat'] = train_data['Age_cat'].fillna(5)

train_data.loc[(train_data.RoomService < 5000),  'RoomService_cat'] = 1
train_data.loc[(train_data.RoomService >= 5000) & (train_data.RoomService < 10000),  'RoomService_cat'] = 2
train_data.loc[(train_data.RoomService >= 10000),  'RoomService_cat'] = 3

train_data = train_data.drop('RoomService',axis=1)


train_data.loc[(train_data.FoodCourt < 10000),  'FoodCourt_cat'] = 1
train_data.loc[(train_data.FoodCourt >= 10000) & (train_data.FoodCourt < 20000),  'FoodCourt_cat'] = 2
train_data.loc[(train_data.FoodCourt >= 20000),  'FoodCourt_cat'] = 3

train_data = train_data.drop('FoodCourt',axis=1)


train_data.loc[(train_data.ShoppingMall < 10000),  'ShoppingMall_cat'] = 1
train_data.loc[(train_data.ShoppingMall >= 10000) & (train_data.ShoppingMall < 20000),  'ShoppingMall_cat'] = 2
train_data.loc[(train_data.ShoppingMall >= 20000),  'ShoppingMall_cat'] = 3

train_data = train_data.drop('ShoppingMall',axis=1)


train_data.loc[(train_data.Spa < 10000),  'Spa_cat'] = 1
train_data.loc[(train_data.Spa >= 10000) & (train_data.Spa < 20000),  'Spa_cat'] = 2
train_data.loc[(train_data.Spa >= 20000),  'Spa_cat'] = 3

train_data = train_data.drop('Spa',axis=1)


train_data.loc[(train_data.VRDeck < 10000),  'VRDeck_cat'] = 1
train_data.loc[(train_data.VRDeck >= 10000) & (train_data.VRDeck < 20000),  'VRDeck_cat'] = 2
train_data.loc[(train_data.VRDeck >= 20000),  'VRDeck_cat'] = 3

train_data = train_data.drop('VRDeck',axis=1)
```

```python
train_data.info()
```

```python
train_data[['Deck', 'Num', 'Side']] = train_data['Cabin'].str.split('/', expand = True) # split cabin on '/'

train_data = train_data.drop(['Cabin'], axis=1) # drop this column
```

```python
#Deck
Deck_trf=['B', 'F', 'A', 'G', 'E', 'D', 'C', 'T', 'other']
Deck=OrdinalEncoder(categories=[Deck_trf])
Deck.fit_transform(train_data[["Deck"]])

train_data[["Deck"]]=Deck.fit_transform(train_data[["Deck"]])

#Num
train_data = train_data.drop(['Num'], axis=1)

#Side
train_data['Side'] = train_data['Side'].fillna('other')

Side_trf=['P', 'S', 'other']
Side=OrdinalEncoder(categories=[Side_trf])
Side.fit_transform(train_data[["Side"]])

train_data[["Side"]]=Side.fit_transform(train_data[["Side"]])
```

```python
train_data[['PassengerId_1', 'PassengerId_2']] = train_data['PassengerId'].str.split('_', expand = True) # split cabin on '/'

train_data = train_data.drop(['PassengerId'], axis=1) # drop this column
```
