# Reading the Data

```python
Data = pd.read_csv("../input/spaceship-titanic/train.csv")
```

```python
Data
```

```python
Data.shape
```

# Drop unnessary columns

```python
dropcolumns = ['RoomService','Spa','ShoppingMall','FoodCourt','Cabin','CryoSleep','VRDeck']
Data = Data.drop(dropcolumns,axis = 1)
```

```python
Data
```

# Inspect the percentage of null values in the columns

```python
round(100*(Data.isnull().sum()/len(Data.index)),2)
```

```python
Data.shape
```

# Drop the rows having greater than or equal to 5 null values

```python
Data = Data[Data.isnull().sum(axis = 1) < 5]
```

```python
Data
```

# Here we see the home planet has Null value

```python
Data['HomePlanet'].describe()
```

# Replace the Null values of HomePlanet with Earth

```python
Data.HomePlanet.fillna("Earth",inplace = True)
```

```python
round(100*(Data.isnull().sum()/len(Data.index)),2)
```

# fill the null values of destination with TRAPPIST-1e

```python
Data.Destination.fillna("TRAPPIST-1e",inplace = True)
```

```python
round(100*(Data.isnull().sum()/len(Data.index)),2)
```

```python
bins = [0,5,15,30,60,100]
labels = ['Kid','Toddler','Teen','Adult','SeniorCitizen']
Data['Age_group'] = pd.cut(Data['Age'],bins = bins,labels = labels,right = False)
```

# grouping the ages by their age-group

```python
Data
```

```python
Data["Age_group"].describe()
```

```python
round(100*(Data.isnull().sum()/len(Data.index)),2)
```

# Filling the nullvalues in the Agegroup with most frequently occurring Values

```python
Data.Age_group.fillna("Teen",inplace = True)
```

```python
Data['VIP'].describe()
```

# Filling the null values in VIP column

```python
Data.VIP.fillna(False,inplace = True)
```

```python
Data
```

# Converting the VIP colums to TRUE --> 1
#                              False -->0

```python
Data.VIP = Data.VIP.map({False:0,True:1})
```
