```python
TrainData['Age']=TrainData['Age'].fillna(TrainData['Age'].mean())
TrainData['Spa']=TrainData['Spa'].fillna(TrainData['Spa'].mean())
TrainData['VRDeck']=TrainData['VRDeck'].fillna(TrainData['VRDeck'].mean())
TrainData['ShoppingMall']=TrainData['ShoppingMall'].fillna(TrainData['ShoppingMall'].mean())
TrainData['FoodCourt']=TrainData['FoodCourt'].fillna(TrainData['FoodCourt'].mean())
TrainData['RoomService']=TrainData['RoomService'].fillna(TrainData['RoomService'].mean())
TrainData['Name']=TrainData['Name'].fillna(TrainData['Name'].mode())
TrainData['Cabin']=TrainData['Cabin'].fillna(TrainData['Cabin'].mode())
TrainData['HomePlanet']=TrainData['HomePlanet'].fillna(TrainData['HomePlanet'].mode())
TrainData['CryoSleep']=TrainData['CryoSleep'].fillna(TrainData['CryoSleep'].mode())
TrainData['Destination']=TrainData['Destination'].fillna(TrainData['Destination'].mode())

testData['Age']=testData['Age'].fillna(testData['Age'].mean())
testData['Spa']=testData['Spa'].fillna(testData['Spa'].mean())
testData['VRDeck']=testData['VRDeck'].fillna(testData['VRDeck'].mean())
testData['ShoppingMall']=testData['ShoppingMall'].fillna(testData['ShoppingMall'].mean())
testData['FoodCourt']=testData['FoodCourt'].fillna(testData['FoodCourt'].mean())
testData['RoomService']=testData['RoomService'].fillna(testData['RoomService'].mean())
testData['Name']=testData['Name'].fillna(testData['Name'].mode())
testData['Cabin']=testData['Cabin'].fillna(testData['Cabin'].mode())
testData['HomePlanet']=testData['HomePlanet'].fillna(testData['HomePlanet'].mode())
testData['CryoSleep']=testData['CryoSleep'].fillna(testData['CryoSleep'].mode())
testData['Destination']=testData['Destination'].fillna(testData['Destination'].mode())
```

```python
HomePlanet=pd.get_dummies(TrainData["HomePlanet"])
HomePlanet2=pd.get_dummies(testData["HomePlanet"])
HomePlanet
```

```python
cryosleep=pd.get_dummies(TrainData["CryoSleep"],drop_first=True)
cryosleep2=pd.get_dummies(testData["CryoSleep"],drop_first=True)
cryosleep
```

```python
Transported=pd.get_dummies(TrainData["Transported"],drop_first=True)
Transported
```

```python
Destination=pd.get_dummies(TrainData["Destination"])
Destination2=pd.get_dummies(testData["Destination"])
Destination
```

```python
Vip=pd.get_dummies(TrainData["VIP"],drop_first=True)
Vip2=pd.get_dummies(testData["VIP"],drop_first=True)
Vip
```

```python
TrainData['CryoSleep']=cryosleep
TrainData['VIP']=Vip
TrainData['Transported']=Transported
TrainData=pd.concat([TrainData,HomePlanet,Destination],axis=1)
TrainData.drop(['PassengerId','HomePlanet','Cabin','Destination','Name'],axis=1,inplace=True)

testData['CryoSleep']=cryosleep2
testData['VIP']=Vip2
testData=pd.concat([testData,HomePlanet2,Destination2],axis=1)
PassengerId=testData['PassengerId']
testData.drop(['PassengerId','HomePlanet','Cabin','Destination','Name'],axis=1,inplace=True)
```
