```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import plotly.offline as py
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.tools as tls
import plotly.figure_factory as ff
py.init_notebook_mode(connected=True)

from autogluon.tabular import TabularPredictor
```

```python
DATA_DIR = '../input/spaceship-titanic/'
train_data = pd.read_csv(DATA_DIR+'train.csv')
test_data = pd.read_csv(DATA_DIR+'test.csv')
submission = pd.read_csv(DATA_DIR+'sample_submission.csv')
label_col = 'Transported'
eval_metric = 'accuracy'
```

```python
train_data.head()
```

```python
def target_count(df):
    trace = go.Bar( x = df['Transported'].value_counts().values.tolist(), 
                    y = ['False','True' ], 
                    orientation = 'h', 
                    text=df['Transported'].value_counts().values.tolist(), 
                    textfont=dict(size=15),
                    textposition = 'auto',
                    opacity = 0.8,marker=dict(
                    color=['skyblue', 'gold'],
                    line=dict(color='#000000',width=1.5)))

    layout = dict(title =  'Target feature')

    fig = dict(data = [trace], layout=layout)
    py.iplot(fig)
```

```python
target_count(df = train_data)
```

```python
predictor = TabularPredictor(label=label_col, eval_metric=eval_metric, verbosity=2).fit(
    train_data, presets='best_quality', time_limit=3600
)

results = predictor.fit_summary()
```

```python
test_pred = predictor.predict(test_data)
```

```python
submission['Transported'] = test_pred
submission.to_csv('TabularAutoML.csv', index=False)
submission
```
