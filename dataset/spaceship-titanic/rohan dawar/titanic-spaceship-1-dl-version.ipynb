```python
# Function Defination
def Univariate(df, feature):
    if (df[feature].nunique())<20:
        print("The no of Unique values in {} are {}".format(feature,df[feature].nunique()))
        sns.countplot(x=feature, data=df)
        print('Dtype:', df[feature].dtypes)
        #plt.show()
    else:
        print("The no of Unique values in {} are {}".format(feature,df[feature].nunique()))
        print('Dtype:', df[feature].dtypes)
    
    
def Analysis(train, test, feature):
    if (train[feature].nunique())<20:
        plt.figure(figsize=(10,4.8))
        plt.subplot(1,2,1)
        sns.countplot(x=feature, data=train)
        plt.title('Distribution of {} in Train set '.format(feature) )
        plt.subplot(1,2,2)
        sns.countplot(x=feature, data=test)
        plt.title('Distribution of {} in Test set '.format(feature) )
        print('The dtype of {} in Training set is {} '.format(feature, train[feature].dtype))
        print('The dtype of {} in Testing set is {}'.format(feature, train[feature].dtype))
        print('The no of records missing in training set is {}, which is {:.2f}% '.format(train[feature].isnull().sum(), (train[feature].isnull().sum()/len(train[feature])*100)))
        print('The no of records missing in testing set is {}, which is {:.2f}% '.format(test[feature].isnull().sum(), (test[feature].isnull().sum()/len(test[feature])*100)))
    else:
        print("The no of Unique values in training set{} are {}".format(feature,train[feature].nunique()))
        print("The no of Unique values in testing set{} are {}".format(feature,train[feature].nunique()))
        print('The dtype of {} in Training set is {} '.format(feature, train[feature].dtype))
        print('The dtype of {} in Testing set is {}'.format(feature, train[feature].dtype))
        print('The no of records missing in training set is {}, which is {:.2f}% '.format(train[feature].isnull().sum(), (train[feature].isnull().sum()/len(train[feature])*100)))
        print('The no of records missing in testing set is {}, which is {:.2f}% '.format(test[feature].isnull().sum(), (test[feature].isnull().sum()/len(test[feature])*100)))

    
def Imputation(df, feature, strategy):
    if strategy == 'mean':
        df[feature] = df[feature].fillna(df[feature].mean())
    elif strategy == 'mode':
        df[feature] = df[feature].fillna(df[feature].mode()[0])
    else:
        df[feature] = df[feature].fillna(df[feature].median())
        
    
    
```

```python
traindf.drop(['Cabin', 'Name', 'PassengerId'], axis=1, inplace=True)
testdf.drop(['Cabin', 'Name'], axis=1, inplace=True)
traindf.shape, testdf.shape
```

```python
# Define Y variable
y_train = traindf['Transported']

features = ["CryoSleep", "Age", "Destination", "RoomService", "FoodCourt", "ShoppingMall", "Spa", "VRDeck", "VIP", "Spa"]

# Get Dummy variable creation.
X_train = pd.get_dummies(traindf[features])
X_test = pd.get_dummies(testdf[features])

# Scaling the features
ss = StandardScaler()
X_train_scaled = ss.fit_transform(X_train)
X_test_scaled = ss.transform(X_test)
```

```python
#Define early Stopping
es = tf.keras.callbacks.EarlyStopping(patience=10,min_delta=0.5,verbose=1)
```

```python
# Model Defination
opt = keras.optimizers.Adam(learning_rate=0.001)
model = keras.models.Sequential()
model.add(keras.layers.Input(shape=X_train.shape[1]))
model.add(keras.layers.Dense(30, activation='relu'))
model.add(keras.layers.Dense(30, activation='tanh'))
model.add(keras.layers.Dense(30, activation='relu'))
model.add(keras.layers.Dense(30, activation='tanh'))
model.add(keras.layers.Dense(30, activation='relu'))
model.add(keras.layers.Dense(20, activation='tanh'))
model.add(keras.layers.Dense(1, activation='sigmoid'))
model.compile(loss ="binary_crossentropy", optimizer=opt, metrics=['accuracy'])
```

```python
# Fit the model
history = model.fit(X_train_scaled,y_train,epochs=100,validation_split=0.1,verbose=1, callbacks=[es])
```

```python
# Make the Prediction
predictions = model.predict(X_test_scaled).round().astype(bool)
```

```python
#Submittion of results.

test = pd.read_csv("/kaggle/input/spaceship-titanic/test.csv")
output = pd.DataFrame(predictions, columns=['Transported'])
output.insert(0,'PassengerId',test['PassengerId'])
output.to_csv('submission.csv', index=False)
```
