<p class='text-primary'> Few columns are of float32 datatype. On conversion to string datatype, missing values are automatically replaced as 'nan'. These values are replaced again as missing values using (np.nan) option. </p>

```python
train['Side']=train['Side'].replace('nan', np.nan)
train['Deck']=train['Deck'].replace('nan', np.nan)
train['Num']=train['Num'].replace('nan', np.nan)
train['Cabin']=train['Cabin'].replace('nan', np.nan)
```

```python
pd.DataFrame(train.isna().sum()).T
```

```python
pd.DataFrame((train.isna().sum()/train.shape[0])*100).T
```

<h2>Approach:</h2>
<p class='text-primary'>1. First, missing values of categorical features are addressed.</p>
   
<p class='text-primary'> 2. Numerical features are addressed using mean/mode values obtained by group_by option in python using different categorical characteristics.</p>

 <p class='text-primary'>3. Families tend to travel to the same destination together. It is assumed that they will do booking on the same deck and likely to have to same side.</p>

<p class='text-primary'>4. Another column 'LastName' is extracted from the 'Name' column. </p>

<h4>Name</h4>

<p class='text-primary'> Before addressing the categorical columns, missing values in the 'Name' column are addressed.</p>

```python
train['Name']=train['Name'].astype(str)
```

```python
train['LastName']=train['Name'].apply(lambda x: (x.split(' ',1)[-1]))
train['LastName']=train['LastName'].replace('nan', np.nan)
```

```python
pd.DataFrame(train.groupby(['Destination', 'HomePlanet', 'CryoSleep', 'Deck'])['LastName'].value_counts())
```

<p class='text-danger'>The idea behind imputation is as follows:</p>
    
<p class='text-primary'> 1. Most frequent last names are found for the groups having same 'destination', 'homeplanet', 'cryosleep' and 'Deck'.</p>

```python
test['CryoSleep']=test['CryoSleep'].astype('float32')
test['VIP']=test['VIP'].astype('float32')

test['Cabin']=test['Cabin'].astype(str)
test['Deck']=test['Cabin'].apply(lambda x: x.split('/',1)[0])
test['Num']=test['Cabin'].apply(lambda x: (x.split('/', 1)[-1]).split('/',1)[0])
test['Side']=test['Cabin'].apply(lambda x: (x.split('/', 1)[-1]).split('/',1)[-1])

test['Name']=test['Name'].astype(str)

test['LastName']=test['Name'].apply(lambda x: (x.split(' ',1)[-1]))
test['LastName']=test['LastName'].replace('nan', np.nan)

# Missing Data Imputation 

#Last Name
test['LastName'] = test['LastName'].fillna(test.groupby(['Destination','HomePlanet','CryoSleep','Deck'])['LastName'].transform(lambda x: x.value_counts().idxmax()))
test['LastName'] = test['LastName'].fillna(test.groupby(['Destination','HomePlanet','CryoSleep'])['LastName'].transform(lambda x: x.value_counts().idxmax()))
test['LastName'] = test['LastName'].fillna(test.groupby(['Destination','HomePlanet'])['LastName'].transform(lambda x: x.value_counts().idxmax()))
test['LastName'] = test['LastName'].fillna(test.groupby(['HomePlanet'])['LastName'].transform(lambda x: x.value_counts().idxmax()))
test['LastName'] = test['LastName'].fillna(test.groupby(['Destination'])['LastName'].transform(lambda x: x.value_counts().idxmax()))

# Home Planet
test['HomePlanet'] = test['HomePlanet'].fillna(test.groupby(['Destination','CryoSleep','Deck'])['HomePlanet'].transform(lambda x: x.value_counts().idxmax()))
test['HomePlanet']=test['HomePlanet'].fillna(test.groupby(['Destination', 'CryoSleep'])['HomePlanet'].transform(lambda x: x.value_counts().idxmax()))
test['HomePlanet']=test['HomePlanet'].fillna(test.groupby(['Destination'])['HomePlanet'].transform(lambda x: x.value_counts().idxmax()))
test['HomePlanet']=test['HomePlanet'].fillna(test['HomePlanet'].mode()[0])

# Destination
#test['Destination'] = test['Destination'].fillna(test.groupby(['HomePlanet','CryoSleep','Deck'])['Destination'].transform(lambda x: x.value_counts().idxmax()))
test['Destination'] = test['Destination'].fillna(test.groupby(['HomePlanet','CryoSleep'])['Destination'].transform(lambda x: x.value_counts().idxmax()))
test['Destination'] = test['Destination'].fillna(test.groupby(['HomePlanet'])['Destination'].transform(lambda x: x.value_counts().idxmax()))

# CryoSleep
test['CryoSleep'] = test['CryoSleep'].fillna(test.groupby(['Destination', 'HomePlanet'])['CryoSleep'].transform(lambda x: x.value_counts().idxmax()))

# Num, Side and Deck
test['Side']=test['Side'].replace('nan', np.nan)
test['Deck']=test['Deck'].replace('nan', np.nan)
test['Cabin']=test['Cabin'].replace('nan', np.nan)

test['Deck'] = test['Deck'].fillna(test.groupby(['Destination', 'HomePlanet', 'CryoSleep'])['Deck'].transform(lambda x: x.value_counts().idxmax()))
test['Side'] = test['Side'].fillna(test.groupby(['Destination', 'HomePlanet', 'CryoSleep'])['Side'].transform(lambda x: x.value_counts().idxmax()))
test['Num'] = test['Num'].fillna(test.groupby(['Destination', 'HomePlanet', 'CryoSleep'])['Num'].transform(lambda x: x.value_counts().idxmax()))

# VIP
test['VIP'] = test['VIP'].fillna(test.groupby(['HomePlanet', 'Destination', 'CryoSleep'])['VIP'].transform(lambda x: x.value_counts().idxmax()))


#Numerical features
cols=['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']
for i in cols:
    test[i] = test[i].fillna(test.groupby(['HomePlanet', 'Destination', 'CryoSleep', 'VIP', 'Deck', 'Side'])[i].transform('median'))

for i in cols:
    test[i] = test[i].fillna(0)

# Age
test['Age'] = test['Age'].fillna(test.groupby(['HomePlanet', 'Destination', 'CryoSleep', 'VIP', 'Deck', 'Side'])['Age'].transform('median'))
test['Age'] = test['Age'].fillna(test['Age'].median())

# Feature Creation

test['Total_expenses']= test['VRDeck']+test['Spa']+test['ShoppingMall']+test['FoodCourt']+test['RoomService']
deck={'A':7,'B':6,'C':5,'D':4,'E':3,'F':2,'G':1, 'T':0}
test=test.replace({'Deck': deck})
test['Deck']=test['Deck'].astype('float32')

test=test[['Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck', 'Total_expenses','Deck', 'CryoSleep', 'VIP', 'Side', 'HomePlanet', 'Destination' ]]
test['Side']=test['Side'].apply(lambda x: 1 if x=='S' else 0)
test=pd.get_dummies(test, drop_first=True)

skewed_columns=['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck', 'Total_expenses']
from scipy.special import boxcox1p
lam=0.15
for i in skewed_columns:
    test[i]= boxcox1p(test[i],lam)
```
