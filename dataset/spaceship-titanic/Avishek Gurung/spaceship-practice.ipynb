```python
train_data = pd.read_csv('/kaggle/input/spaceship-titanic/train.csv')
test_data = pd.read_csv('/kaggle/input/spaceship-titanic/test.csv')

missing_passenger_ids = test_data.pop('PassengerId')

category_columns = ['HomePlanet', 'Destination', 'CryoSleep', 'VIP']
train_data[category_columns]=train_data[category_columns].fillna(train_data.mode().iloc[0])
test_data[category_columns]=test_data[category_columns].fillna(test_data.mode().iloc[0])

columns = ['Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']
train_data[columns] = train_data[columns].fillna(train_data.mode().iloc[0])
test_data[columns] = test_data[columns].fillna(test_data.mode().iloc[0])

train_data['berth'] = train_data['Cabin'].str.split("/",0).str[2]
test_data['berth'] = test_data['Cabin'].str.split("/",0).str[2]

train_data = train_data.drop(['Cabin','Name', 'PassengerId'], axis=1)
test_data = test_data.drop(['Cabin','Name'], axis=1)

train_data = pd.get_dummies(train_data, drop_first=True)
test_data = pd.get_dummies(test_data, drop_first=True)


Y_train = train_data.pop('Transported')
X_train = train_data
X_test = test_data

from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(max_iter=7000)
log_reg.fit(X_train, Y_train)
Y_predict = log_reg.predict(X_test)

submission = pd.DataFrame({'PassengerId': missing_passenger_ids, 'Transported': Y_predict})
submission.head()
submission.to_csv("submission.csv",index=False)
submission.shape
```
