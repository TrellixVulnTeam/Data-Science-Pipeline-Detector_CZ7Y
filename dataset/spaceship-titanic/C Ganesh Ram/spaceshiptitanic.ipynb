```python
import numpy as np
```

```python
def detect_outlier(data):
    threshold=3.5
    mean = np.mean(data)
    std = np.std(data)
    outliers = []
    for y in data:
        z = (y - mean)/std 
        if np.abs(z) > threshold:
            outliers.append(y)
    return outliers
```

```python

out1=detect_outlier(data.FoodCourt)
out2=detect_outlier(data.ShoppingMall)
out3=detect_outlier(data.Spa)
out4=detect_outlier(data.VRDeck)
print(len(out1),len(data.FoodCourt))
print(len(out2),len(data.ShoppingMall))
print(len(out3),len(data.Spa))
print(len(out4),len(data.VRDeck))
```
