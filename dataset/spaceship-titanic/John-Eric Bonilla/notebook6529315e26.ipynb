```python
df_feat = pd.DataFrame(scaled_features,columns=df_new.columns[:-1])
df_feat.head()
```

```python
X = df_feat
```

```python
y = df_new['Transported']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.1, random_state=5)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train[:10]
```

```python
len(X_train.columns)
```

```python
model = keras.Sequential([
    keras.layers.Dense(32, input_shape=(16,), activation='relu'),
    keras.layers.Dense(16, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid'),
])

model.compile(optimizer='adam',
             loss='binary_crossentropy',
             metrics=['accuracy'])

model.fit(X_train, y_train, epochs=36)
```

```python
model.summary()
```

```python
model.evaluate(X_test, y_test)
```

```python
yp = model.predict(X_test)
yp[:5]
```

```python
y_test[:5]
```

```python
y_pred = []
for element in yp:
    if element > 0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
y_pred[5]
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(classification_report(y_test, y_pred))
```

###### Confusion Matrix

```python
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_pred)

plt.figure(figsize=(16,16))
sns.heatmap(cm,cmap='plasma_r', annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Actual')
```
