# <span style='color:#A80808'>Objective</span>

Compare different classifiers for a better model selection.



```python
import warnings
warnings.simplefilter('ignore')

import sys
import pandas as pd

# Pycaret is a low code solution for model comparison, tuning and blending
# Link to the project: https://pycaret.gitbook.io/docs/
sys.path.append('../input/pycaret/pycaret')
sys.path.append('../input/pycaret/pyod')
from pycaret.classification import *
```

# <span style='color:#A80808'>Data</span>

```python
train = pd.read_csv('../input/spaceship-titanic-feature-engineering/train.csv')
train.head()
```

# <span style='color:#A80808'>Model </span>

```python
s = setup(data = train,
          target = 'Transported',
          silent = True)
```

```python
model = create_model('lightgbm')
```

```python
model = tune_model(model, n_iter=10000, optimize='Accuracy')
```

```python
model
```

```python
model = finalize_model(model)
```

```python
test = pd.read_csv('../input/spaceship-titanic-feature-engineering/test.csv')
pred = predict_model(model, data=test)
```

```python
submission = pd.read_csv('../input/spaceship-titanic/sample_submission.csv')
submission.Transported = pred.Label
submission.to_csv('submission.csv', index=False)
submission.head()
```
