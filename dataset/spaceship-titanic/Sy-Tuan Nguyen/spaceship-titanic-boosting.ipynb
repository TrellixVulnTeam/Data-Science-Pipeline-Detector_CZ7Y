This notebook presents a simple ensemble model using predicted results from the most relevant public notebooks. 

```python
import os
import numpy as np
import pandas as pd
from scipy import stats
import plotly.express as px
```

```python
targetName = 'Transported'
competitionDir = '/kaggle/input/spaceship-titanic'
submission = pd.read_csv('../input/spaceship-titanic/sample_submission.csv')
train = pd.read_csv('../input/spaceship-titanic/train.csv')
```

```python
preds = []
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        if (dirname != competitionDir) & ('.csv' in filename):
            df = pd.read_csv(os.path.join(dirname, filename))
            if (df.shape==submission.shape):
                print(os.path.join(dirname, filename))
                preds.append(df[targetName])
```

```python
submission[targetName] = stats.mode(np.array(preds), axis=0)[0].transpose()
submission.to_csv("submission.csv", index=False)
```

```python
test_target_df = pd.DataFrame((submission[targetName].value_counts())).reset_index()
test_target_df.columns = [targetName, 'count']
fig = px.bar(data_frame = test_target_df, 
             x = targetName,
             y = 'count' , 
             color = "count",
             color_continuous_scale="Emrld", title='Test') 
fig.show()

train_target_df = pd.DataFrame((train[targetName].value_counts())).reset_index()
train_target_df.columns = [targetName, 'count']
fig = px.bar(data_frame = train_target_df, 
             x = targetName,
             y = 'count' , 
             color = "count",
             color_continuous_scale="Emrld", title='Train') 
fig.show()
```
