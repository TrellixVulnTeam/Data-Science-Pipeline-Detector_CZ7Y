```python
train = pd.read_csv("../input/spaceship-titanic/train.csv")
test = pd.read_csv("../input/spaceship-titanic/test.csv")
```

```python
from supervised.preprocessing.eda import EDA
```

```python
y_train = train["Transported"]
X_train = train.drop(["Transported"], axis=1)
```

I use extensive_eda to obtain some graphs which permit to evaluate the relevance of each critera

```python
EDA.extensive_eda(X_train,y_train,save_path=".")
```

```python
automl = AutoML(mode="Compete", eval_metric="accuracy") 
```

I choose to use the compete mode, because Optima mode predicts 6 hours to find a better solution

```python
automl.fit(X_train, y_train)
```

```python
automl.report()
```

```python
predictions = automl.predict(test)
```

```python
sample_submission = pd.read_csv('../input/spaceship-titanic/sample_submission.csv')
```

```python
sample_submission['Transported'] = predictions
```

```python
sample_submission["Transported"] = sample_submission["Transported"].map({0:False , 1:True})
sample_submission
```
