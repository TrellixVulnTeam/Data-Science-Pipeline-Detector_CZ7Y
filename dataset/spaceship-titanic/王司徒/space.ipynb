```python
train_data = pd.read_csv('/kaggle/input/spaceship-titanic/train.csv')
test_Data =pd.read_csv('/kaggle/input/spaceship-titanic/test.csv')
sample = pd.read_csv('/kaggle/input/spaceship-titanic/sample_submission.csv')
train_data['HomePlanet'].value_counts()
train_y = train_data['Transported'].replace(['True','False'],[1,0]).astype("int32")
train_data=train_data.drop('Transported',axis=1)
```

```python
data = pd.concat([train_data,test_Data],axis=0)
data
```

```python


dummy_list = []
data=data.drop(['Name','PassengerId'],axis=1)
columns_list=data.columns

```

```python
for name in columns_list:
    if data[name].dtypes == 'object':
        dummy_list.append(name)

data2 = data[dummy_list]

data=pd.concat([data,pd.get_dummies(data2)],axis=1)
data=data.drop(dummy_list,axis=1)
```

```python

train=data.iloc[:8693,:]
test=data.iloc[8693:,:]
print(train.shape,test.shape)
```

```python
from sklearn.model_selection import  train_test_split 

from catboost import CatBoostClassifier
train_X,test_X,train_y,test_y=train_test_split(train,train_y,random_state=33,test_size=0.3)


```

```python
from sklearn.metrics import classification_report
clf=CatBoostClassifier()
clf.fit(train_X, train_y)
result=clf.predict(test_X)
print(classification_report(result,test_y))
```

```python
test_result=clf.predict(test)
```

```python
test_result=pd.DataFrame(test_result,columns=['Transported'])
test_result=test_result.replace([0,1],['False','True'])
```
