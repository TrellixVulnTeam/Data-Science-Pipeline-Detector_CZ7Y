# Preprocessing

```python
train.drop(['PassengerId', 'Name'], inplace = True, axis = 1)
train.columns
```

```python
test_id = test['PassengerId']
test.drop(['PassengerId', 'Name'], inplace = True, axis = 1)
test.columns
```

```python
train.isnull().sum()
```

```python
test.isnull().sum()
```

```python
# Findning null values in training and testing data.

for i in train.columns:
    if train[i].isnull().sum().any():
        print("Training")
        print(train[i].value_counts())
        print("\n")
        print("Testing")
        print(test[i].value_counts())
        print("*********"*5)
        print()
```

**For most of the collumns we will be replacing with there mode and for age we will be replacing with mean.**

```python
# Filling null values with mean for Age and with mode for rest of the columns.

for i in train.columns:
    if train[i].isnull().sum().any():
        if i == 'Age':
            train[i] = train[i].fillna(train[i].mean())
            test[i] = test[i].fillna(test[i].mean())
        else:
            train[i] = train[i].fillna(train[i].mode()[0])
            test[i] = test[i].fillna(test[i].mode()[0])
```

```python
print("Null values in training dataset = {}".format(train.isnull().sum().any()))
print("Null values in testing dataset = {}".format(test.isnull().sum().any()))
```

```python
# Function for dividing cabinate name in First name, Middle name and Last name.

def Cabin_Transform(data):
    cabin = list(data)
    cabin_First = []
    cabin_Middle = []
    cabin_Last = []
    for i in cabin:
        l = str(i).split("/")
        cabin_First.append(l[0])
        cabin_Middle.append(l[1])
        cabin_Last.append(l[2])
    return cabin_First, cabin_Middle, cabin_Last
```

```python
# Adding different columns in traning and testing data for cabinate names.

cabin_list = [train['Cabin'], test['Cabin']]
for i in range (len(cabin_list)):
    cabin_First, cabin_Middle, cabin_Last = Cabin_Transform(cabin_list[i])
    if i==0:
        train.drop('Cabin', inplace = True, axis = 1)
        train['Cabin_First'] = cabin_First
        train['Cabin_Middle'] = cabin_Middle
        train['Cabin_Last'] = cabin_Last
    else:
        test.drop('Cabin', inplace = True, axis = 1)
        test['Cabin_First'] = cabin_First
        test['Cabin_Middle'] = cabin_Middle
        test['Cabin_Last'] = cabin_Last
```

```python
train.head()
```

```python
test.head()
```

```python
# Changning datatype to int.

train['Cabin_Middle'] = train['Cabin_Middle'].astype('int')
test['Cabin_Middle'] = test['Cabin_Middle'].astype('int')
```

```python
train.info()
```

```python
test.info()
```

```python
# Converting object type columns to integer or float type.

la = LabelEncoder()
for i in test.columns:
    if train[i].dtype == 'object' or train[i].dtype == 'bool':
        train[i] = la.fit_transform(train[i])
        test[i] = la.fit_transform(test[i])
train['Transported'] = la.fit_transform(train['Transported'])
```

# Random Forest + Randomized Search

```python
params = {
    "n_estimators": [i for i in range (10, 800, 10)], 
    "max_depth": [i for i in range (2, 40, 2)],
    "min_samples_leaf" : [i for i in range (15, 40)],
    "min_samples_split" : [i for i in range (5,240)]
}
rcla = RandomForestClassifier()
rcla
```

```python
grid = RandomizedSearchCV(estimator = rcla, param_distributions = params, cv = 5, scoring = 'accuracy', n_jobs = -1)
grid.fit(xtrain, ytrain)
```

```python
best_param_rf = grid.best_params_
best_param_rf
```

```python
grid.best_score_
```

```python
rcla = RandomForestClassifier(**best_param_rf)
rcla.fit(xtrain, ytrain)
```

```python
predicted = rcla.predict_proba(xtrain)[:,1]
train_score = accuracy_score(predicted.round(), ytrain)*100
print("Accuracy using Random Forest on training data is {} %".format(train_score))
```

```python
predicted = rcla.predict_proba(xtest)[:, 1]
test_score = accuracy_score(predicted.round() , ytest)*100
print("Accuracy using Random Forest testing data is {} %".format(test_score))
```

```python
predicted.round()
```

```python
mae = mean_squared_error(predicted, ytest)
mae
```

```python
F1_Score = f1_score(predicted.round(), ytest)
F1_Score
```
