```python
import pandas as pd
# data reading for learning
train_data = pd.read_csv('/kaggle/input/spaceship-titanic/train.csv')
print(train_data.head())
print(train_data.isna().sum())
```

```python
# data preprocessing
train_data = train_data.fillna(0)
train_data.PassengerId = train_data.PassengerId.astype('category').cat.codes
train_data.HomePlanet = train_data.HomePlanet.astype('category').cat.codes
train_data.Cabin = train_data.Cabin.astype('category').cat.codes
train_data.Destination = train_data.Destination.astype('category').cat.codes
train_data.Name = train_data.Name.astype('category').cat.codes
print(train_data.head())
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(train_data, shuffle=False)
x_train = train.drop('Transported', axis=1) 
y_train = train['Transported']
x_test =  test.drop('Transported', axis=1)
y_test = test['Transported']
```

```python
# learning
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_jobs=-1, min_samples_leaf=10, n_estimators=100)
model.fit(x_train, y_train)
print(model.score(x_test, y_test))
```

```python
# File reading for predictions
test_data = pd.read_csv('/kaggle/input/spaceship-titanic/test.csv')
test_data_original = test_data
```

```python
# Data preprocessing
test_data = test_data.fillna(0)
test_data.PassengerId = test_data.PassengerId.astype('category').cat.codes
test_data.HomePlanet = test_data.HomePlanet.astype('category').cat.codes
test_data.Cabin = test_data.Cabin.astype('category').cat.codes
test_data.Destination = test_data.Destination.astype('category').cat.codes
test_data.Name = test_data.Name.astype('category').cat.codes
```

```python
result = model.predict(test_data)
```

```python
# Making submission file
test_data_original['Transported'] = result
submission = test_data_original[['PassengerId', 'Transported']]
submission.to_csv('submission.csv', index=False)
```
