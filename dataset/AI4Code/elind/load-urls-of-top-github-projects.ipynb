```python
access_tokens = [
    "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",
    "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",
    "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",
    "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",
    "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",
    "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",
    "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",
    "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",
    "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",
]  # use as many GitHub personal access tokens as possible

n_access_tokens = len(access_tokens)
n_access_tokens
```

```python
language = "python"  # select programming language

for access_token in tqdm(access_tokens):
    print(access_token)
    g = Github(access_token)
    print(
        len(list(g.search_repositories(f"language:{language} stars:{9999}..{10099}")))
    )  # test personal access tokens
    print("\n")
```

```python
n_stars, shift = (
    2 ** 12,
    2 ** 11,
)  # go for lowest possible number of stars and highest shift before getting an error from GitHub
g = Github("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")  # use last personal access token
repositories = list(g.search_repositories(f"language:{language} stars:>={n_stars}"))
n_repositories = len(repositories)
print(f"language:{language} stars:>={n_stars}")
print(n_repositories)
```

```python
i, n_repositories_0, counter = 0, n_repositories, 0

while True:
    try:
        g = Github(access_tokens[i % n_access_tokens])
        repositories += list(
            g.search_repositories(
                f"language:{language} stars:{n_stars-shift}..{n_stars}"
            )
        )
        print(f"language:{language} stars:{n_stars-shift}..{n_stars}")
        n_stars -= shift
        n_repositories = len(repositories)
        print(n_repositories)
        print("\n")
        i += 1
    except:
        i += 1

        try:
            g = Github(access_tokens[i % n_access_tokens])
            repositories += list(
                g.search_repositories(
                    f"language:{language} stars:{n_stars-shift}..{n_stars}"
                )
            )
            print(f"language:{language} stars:{n_stars-shift}..{n_stars}")
            n_stars -= shift
            n_repositories = len(repositories)
            print(n_repositories)
            print("\n")
            i += 1
        except:
            shift = max(1, int(shift * 2 / 3))
            i += 1

    if n_repositories == n_repositories_0:
        counter += 1
    else:
        n_repositories_0 = n_repositories
        counter = 0

    print(counter)
    print("\n")

    if (counter == n_access_tokens) | (n_stars == 100):
        break
```

# Save

```python
urls = [repository.clone_url for repository in repositories]  # get repository urls
df = pd.DataFrame(urls, columns=["url"]).drop_duplicates()
df.reset_index(drop=True, inplace=True)
df.to_csv(f"urls_of_top_{language}_github_projects.csv", index=False)
df
```
