<h1><span class="label label-default" style="background-color:#FFE4C4;border-radius:100px 100px; font-weight: bold; font-family:Garamond; font-size:20px; color:black; padding:10px">What is a monotonic relationship?</span></h1><br>

By Juhi Ramzai

"A monotonic relationship is a relationship that does one of the following:"

(1) as the value of one variable increases, so does the value of the other variable, OR,"

(2) as the value of one variable increases, the other variable value decreases."

"BUT, not exactly at a constant rate whereas in a linear relationship the rate of increase/decrease is constant."

https://towardsdatascience.com/clearly-explained-pearson-v-s-spearman-correlation-coefficient-ada2f473b8#:~:text=What%20is%20a%20monotonic%20relationship,the%20other%20variable%20value%20decreases.

![](https://miro.medium.com/max/1400/1*9-V2iljnY1aM6OgVxHneVg.png)https://medium.com/analytics-vidhya/application-of-monotonic-constraints-in-machine-learning-models-334564bea616

"Application of Monotonic Constraints in Machine Learning Models"

Author: Ajay Tiwari

"Monotonically Increasing Function — A function is called monotonically increasing if, for all x and y such that x≤y one has f(x)≤ f(y), so f preserves the order. This function does not exclusively have to increase, it simply must not decrease."

"Monotonically Decreasing Function — A function is called monotonically decreasing if, whenever x≤y, then f(x)≥f(y), so it reverses the order. This function does not exclusively have to decrease, it simply must not increase."

https://medium.com/analytics-vidhya/application-of-monotonic-constraints-in-machine-learning-models-334564bea616

```python
from sklearn.preprocessing import LabelEncoder

#fill in mean for floats
for c in df.columns:
    if df[c].dtype=='float16' or  df[c].dtype=='float32' or  df[c].dtype=='float64':
        df[c].fillna(df[c].mean())

#fill in -999 for categoricals
df = df.fillna(-999)
# Label Encoding
for f in df.columns:
    if df[f].dtype=='object': 
        lbl = LabelEncoder()
        lbl.fit(list(df[f].values))
        df[f] = lbl.transform(list(df[f].values))
        
print('Labelling done.')
```

```python
import matplotlib
import statsmodels.tsa.api as tsa

%matplotlib inline

matplotlib.style.use("Solarize_Light2")

from pylab import rcParams
import statsmodels.api as sm
```

```python
y = df.loc[~df.id.isnull()][['id', 'cell_order']]

y_log=np.log1p(y)
```

```python
#Code by Oniel Gracious https://www.kaggle.com/onielg/simplecatboostwithclassifierchains/notebook

#histogram and normal probability plot
from scipy import stats

from scipy.stats import norm
target=['cell_order','id']
for t in target:
    fig = plt.figure()
    sns.distplot(y_log[t], fit=norm);
    res = stats.probplot(y_log[t], plot=plt)
```

```python
#Code by Amiya Ranjan Rout https://www.geeksforgeeks.org/python-kendall-rank-correlation-coefficient/

# Import required libraries
from scipy.stats import kendalltau
  
# Taking values from the above example in Lists
X = 'id'
Y = 'cell_order'
  
# Calculating Kendall Rank correlation
corr, _ = kendalltau(X, Y)
print('Kendall Rank correlation: %.5f' % corr)
  
# This code is contributed by Amiya Rout
```
