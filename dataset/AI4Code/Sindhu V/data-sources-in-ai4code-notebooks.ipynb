This is inspired by Y.Nakama's [notebook to identified top referenced Kagglers](https://www.kaggle.com/code/yasufuminakama/ai4code-most-common-kagglers/notebook). This notebook identifies all Kaggle sources used by the notebooks in AI4Code's train dataset. 

Since different competitions' texts can have somewhat similar content (and maybe relatively similar processing, etc), this feature can be used to cluster the notebooks or for stratification purposes. 

There can still be false negatives using this approach. For instance, code cells like this `DATA_DIR = "/kaggle/input"; SOURCE = f"{DATA_DIR}/{DATA_SOURCE}"` still have a data source, but just not captured by the regex I'm using. Feel free to fork and fix it. 

```python
import pandas as pd, numpy as np, json
from tqdm import tqdm 
tqdm.pandas()
df = pd.read_csv("../input/json-files-to-csv/ai4code_train_flat.csv")
```

```python
import re
def get_comp_sources(dat_str):
    itr = re.finditer("/input/([a-z0-9\-]+)/", dat_str.lower())
    return list(set([group.group(1) for group in itr]))
    
df["data_source"] = df["dat"].progress_apply(get_comp_sources)
```

```python
df["total_sources"] = df["data_source"].apply(len)
pd.set_option("display.max_colwidth", None)
display(df[df["total_sources"]>1].sample(20)[["id", "data_source", "total_sources"]])
pd.set_option("display.max_colwidth", 60)
```
