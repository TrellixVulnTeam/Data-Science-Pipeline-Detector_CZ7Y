## Load metric

With `evaluate` anybody can add community metrics (see the [guide](https://huggingface.co/docs/evaluate/creating_and_sharing) in the documentation). We added the AI4Code evaluation metric as a community metric so anybody can easily use it which comes with an interactive widget and a metric card:

https://huggingface.co/spaces/lvwerra/ai4code

Using a community metric with evaluate is just a few lines of code:

```python
import evaluate

ai4code_metric = evaluate.load("lvwerra/ai4code")
```

## Compute result

With `evaluate` you can either pass all predictions/references directly to `compute`:

```python
result = ai4code_metric.compute(
    references=[["a", "b", "c"], ["x", "y"]],
    predictions=[["b", "a", "c"], ["x", "y"]]
)

print(result)
```

Or you can add single samples with `add` (or batches with `add_batch`):

```python
ai4code_metric.add(references=["a", "b", "c"],  predictions=["b", "a", "c"])
ai4code_metric.add(references=["x", "y"],  predictions=["x", "y"])

result = ai4code_metric.compute()
print(result)
```

## Save results

The `evaluate` library comes with a save function so you can also save the results locally as a JSON file with additional useful information for reproducibility:

```python
path = evaluate.save("./result/", **result, **{"model_name": "my_model"})
path
```

```python
import json

with open(path, "r") as f:
    print(json.dumps(json.load(f), indent=4))
```
