```python
import requests
```

Get list of top 5000 Python packages from pypi.org in order to expand semantically notebooks code sections that import them.

```python
packages_list = requests.get("https://hugovk.github.io/top-pypi-packages/top-pypi-packages-30-days.min.json")
```

```python
import json
```

```python
packages_dict = json.loads(packages_list.text)
```

```python
import os
```

```python
from bs4 import BeautifulSoup
```

```python
from tqdm.notebook import tqdm
```

```python
descriptions = []
names = []
downloads = []
for project in tqdm(packages_dict["rows"]):
    project_name = project["project"]
    download_count = project["download_count"]
    project_url = f"https://pypi.org/project/{project_name}"
    project_html = requests.get(project_url)
    soup = BeautifulSoup(project_html.text)
    project_text = soup.get_text()
    descriptions.append(project_text)
    names.append(project_name)
    downloads.append(download_count)
```

```python
import pandas as pd
```

```python
df_packages = pd.DataFrame(data={"name":names, "download_count":downloads, "description":descriptions})
```

```python
df_packages.to_csv("python_packages.csv", index=False)
```
