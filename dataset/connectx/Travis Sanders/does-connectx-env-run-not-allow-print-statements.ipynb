I've implemented a very simple agent for the ConnectX problem, but something bizarre is happening within a nested for-loop.

````python
from kaggle_environments import make

env = make("connectx", debug=True)
````

````python
def my_agent(observation, configuration):
    possible_actions = [i for i in range(7)]

    for my_action in possible_actions:
        for their_action in possible_actions:
            my_score = 50
            their_score = 100
            print(f"{my_action}, {their_action}: my score {my_score}, their score {their_score}")

    return 1
````
