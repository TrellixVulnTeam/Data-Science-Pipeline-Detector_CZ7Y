```python
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
df = pd.read_csv('/kaggle/input/santander-value-prediction-challenge/train.csv')
```

***Linéaire***

```python
from sklearn.linear_model import LinearRegression
```

```python
X = df.drop(['target','ID'],axis=1)
y = df.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=1)
```

```python
lm = LinearRegression()
lm.fit(X_train, y_train)
y_pred = lm.predict(X_test)
```

```python
plt.figure(figsize=(12,12))
plt.scatter(y_test, y_pred)
plt.plot([y_test.min(),y_test.max()],[y_test.min(),y_test.max()], color='red', linewidth=3)
plt.title("Valeur réel de la transaction vs predictions")
plt.xlabel("Valeur")
plt.ylabel("Prediction de valeur")
```

```python
lm.score(X_test,y_test)
```

***Forest***

```python
from sklearn import ensemble
```

```python
X = df.drop(['target','ID'],axis=1)
y = df.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
rf = ensemble.RandomForestRegressor()
rf.fit(X_train, y_train)
y_rf = rf.predict(X_test)
```

```python
plt.figure(figsize=(12,12))
plt.scatter(y_test, y_rf)
plt.plot([y_test.min(),y_test.max()],[y_test.min(),y_test.max()], color='red', linewidth=3)
plt.title("Valeur réel de la transaction vs predictions")
plt.xlabel("Valeur")
plt.ylabel("Prediction de valeur")
```

```python
print(rf.score(X_test,y_test))
```

***xgboost***

```python
import xgboost as XGB
```

```python
xgb = XGB.XGBRegressor()
xgb.fit(X_train, y_train)
y_xgb = xgb.predict(X_test)
```

```python
plt.figure(figsize=(12,12))
plt.scatter(y_test, y_xgb)
plt.plot([y_test.min(),y_test.max()],[y_test.min(),y_test.max()], color='red', linewidth=3)
plt.title("Valeur réel de la transaction vs predictions")
plt.xlabel("Valeur")
plt.ylabel("Prediction de valeur")
```

```python
print(xgb.score(X_test,y_test))
```
