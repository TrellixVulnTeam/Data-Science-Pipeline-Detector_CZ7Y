```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from collections import Counter
# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import os
#for dirname, _, filenames in os.walk('/kaggle/input'):
    #for filename in filenames:
        #print(os.path.join(dirname, filename))

# You can write up to 20GB to the current directory (/kaggle/working/) that gets preserved as output when you create a version using "Save & Run All" 
# You can also write temporary files to /kaggle/temp/, but they won't be saved outside of the current session


# Experiance in tournament (Wins/Loses, participation)
# Conferences in tournament
# 2014 - 2021 Conference tournament winners and losers
Teams = pd.read_csv('/kaggle/input/ncaam-march-mania-2021-spread/MDataFiles_Stage1_Spread/MTeams.csv')


Season_results_detail = pd.read_csv('/kaggle/input/ncaam-march-mania-2021-spread/MDataFiles_Stage2_Spread/MRegularSeasonDetailedResults.csv')
Season_results_detail = Season_results_detail[Season_results_detail['Season'] >2013]
#print(Tourney_results.sort_values(by =['LTeamID']))
Teams.iloc[:,1:4] = 0
Teams[5] =  Teams.iloc[:,1]+Teams.iloc[:,2]
Teams.rename(columns = { 'TeamName': 'Wins','FirstD1Season':'Losses','LastD1Season':'# of Tournament games',5:'Win %'}, inplace = True)
#print(Teams)

FGA_W=Season_results_detail.iloc[:,9]
FGM_W=Season_results_detail.iloc[:,8]
FGA_L=Season_results_detail.iloc[:,22]
FGM_L=Season_results_detail.iloc[:,21]


dict_FGA = Counter(FGA_W) + Counter(FGA_L)
dict_FGM = Counter(FGM_W) + Counter(FGA_L)



Team_wins = Season_results_detail.iloc[:,2]
Team_Losses = Season_results_detail.iloc[:,4]
duplicate_dict_W = Counter(Team_wins)
duplicate_dict_L = Counter(Team_Losses)

##
## Expeince: the amount of wins and losses in the tournament since 2014
##
def Win_p(id):  
    if id in (Season_results_detail.values) :
        win_p=round((duplicate_dict_W[id] /(duplicate_dict_W[id] + duplicate_dict_L[id]))*100,2)
       
    else:
        win_p=0
    return(win_p)


def FG_game(id): #can only get winner FG
    #equation = (FGA+FGM)/#games
    if id in (Season_results_detail.values):
        fg_game=round((dict_FGA[id] +dict_FGM[id])/ (duplicate_dict_W[id] +duplicate_dict_L[id]),2)
    
    else:
        fg_game=0
           
    return(fg_game)


win_p={}
fg_game={}

for i in range (1101,1471):
    win_p[i]=Win_p(i)
    fg_game[i]=FG_game(i)
    
```
