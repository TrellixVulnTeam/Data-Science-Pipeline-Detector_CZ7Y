```python
dataframeequipos=ps.sqldf("""select a.TeamID,a.TeamName,b.ConfAbbrev,b.Season ,  
         num_entrenadores,Description
         from teams a left join
         conferencias b on a.TeamID=b.TeamId 
         inner join 
         (select season,teamid,count(coachname)num_entrenadores from MTeamCoaches  group by 1,2 )d
         on a.Teamid=d.TEamID and b.season=d.season
         inner join Conferences e
         on b.ConfAbbrev=e.ConfAbbrev
 
         """)
```

```python
partidos=ps.sqldf("""select a.Daynum, b.*, c.TeamID as LTeamID,
        c.TeamName as Lteamname,
        c.ConfAbbrev as LConfAbbrev,
        c.num_entrenadores as LNum_entrenadores,
        c.Description as LDesconf,
        WScore-LScore as diferencia,
WScore,LScore,WLoc,NumOT,WFGM,WFGA,LFGA3,LFTM,LFTA,LOR,LDR,LAst,LTO,LStl,LBlk,LPF
        from MRegularSeasonDetailedResults  a 
         inner join dataframeequipos b
         on WteamID=b.TeamID
          and a.Season=b.Season
               inner join dataframeequipos c
         on LteamID=c.TeamID
          and a.Season=c.Season
          
         """)
```

```python
partidos_sin_detalle_regular=ps.sqldf("""select a.Daynum, b.*, c.TeamID as LTeamID,
        c.TeamName as Lteamname,
        c.ConfAbbrev as LConfAbbrev,
        c.num_entrenadores as LNum_entrenadores,
        c.Description as LDesconf,WScore,LScore,
        WScore-LScore as diferencia

        from MRegularSeasonCompactResults  a 
         inner join dataframeequipos b
         on WteamID=b.TeamID
          and a.Season=b.Season
               inner join dataframeequipos c
         on LteamID=c.TeamID
          and a.Season=c.Season
          
         """)
```

```python
partidos_sin_detalle_regular
```

```python
partidos_sin_detalle_totalr=ps.sqldf("""
select a.*,'regular' as phase from 
partidos_sin_detalle_regular a
union all 
select a.*,'torneo' as phase from
partidos_sin_detalle_torneo a
""")
```

```python
partidos_sin_detalle_totalr
```

```python
partidos_sin_detalle_totalr.hist(figsize=(20,20))
plt.show()
```

```python
fig = px.box(partidos_sin_detalle_totalr, x="Season", y="diferencia", color='phase')
fig.show()
```

```python
ps.sqldf("""select Teamid,Lteamid,count(*) from partidos_sin_detalle_totalr
group by 1,2 order by 3 desc""")
```

```python
fig = px.scatter(partidos_sin_detalle_totalr, x="WScore", y="LScore",
            size="diferencia", color="phase",
                 hover_name="Season", log_x=True, size_max=60)
fig.show()
```
