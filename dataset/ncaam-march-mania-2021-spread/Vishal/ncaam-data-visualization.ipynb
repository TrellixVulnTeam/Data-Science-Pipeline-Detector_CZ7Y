```python
import plotly.offline as py
# Add histogram data
x1 = df_regular.loc[df_regular['WLoc'] == 'H', 'ScoreDiff']
x2 = df_regular.loc[df_regular['WLoc'] == 'N', 'ScoreDiff']
x3 = df_regular.loc[df_regular['WLoc'] == 'A', 'ScoreDiff']
# Group data together
hist_data = [x1, x2, x3]
group_labels = ['Home Team', 'Neutral Court', 'Visiting Team']
# Create distplot with custom bin_size
fig = ff.create_distplot(hist_data, group_labels,
                         curve_type='kde', show_hist=False, show_rug=False)
# Add title
fig['layout'].update(title='Scorediff Distribution in Percent by Wloc Status')
fig['layout'].update(xaxis=dict(range=[15, 60], dtick=5))
# Plot
py.iplot(fig, filename='Distplot with Multiple Datasets')
```

# Home Team

```python
df_Season = pd.DataFrame(columns=["Season", "% of Wscore"])
i=0
for field in list(df_regular['Season'].unique()):
    ratio = df_regular[(df_regular['Season']==field)&(df_regular['WLoc']=="H")].shape[0] / df_regular[df_regular['Season']==field].shape[0]
    df_Season.loc[i] = (field, ratio*100)
    i += 1    
df_JS = df_Season.groupby(by="Season").sum()
df_JS.iplot(kind='bar',title='Season (%)')
```

# Neutral Court

```python
df_Season = pd.DataFrame(columns=["Season", "% of Wscore"])
i=0
for field in list(df_regular['Season'].unique()):
    ratio = df_regular[(df_regular['Season']==field)&(df_regular['WLoc']=="N")].shape[0] / df_regular[df_regular['Season']==field].shape[0]
    df_Season.loc[i] = (field, ratio*100)
    i += 1  
df_JS = df_Season.groupby(by="Season").sum()
df_JS.iplot(kind='bar',title='Season (%)', color='red')
```
