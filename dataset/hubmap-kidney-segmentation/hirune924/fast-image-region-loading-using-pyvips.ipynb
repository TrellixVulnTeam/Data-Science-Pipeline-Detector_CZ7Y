# Import libraries

```python
import pyvips
import numpy as np
import matplotlib.pyplot as plt

format_to_dtype = {
    'uchar': np.uint8,
    'char': np.int8,
    'ushort': np.uint16,
    'short': np.int16,
    'uint': np.uint32,
    'int': np.int32,
    'float': np.float32,
    'double': np.float64,
    'complex': np.complex64,
    'dpcomplex': np.complex128,
}
```

# load region

```python
%%timeit
for i in range(5):
    image = pyvips.Image.new_from_file('../input/hubmap-kidney-segmentation/train/0486052bb.tiff')
    patch = image.crop(i*2048, i*2048, 2048, 2048)

    np_img = np.ndarray(buffer=patch.write_to_memory(),
                       dtype=format_to_dtype[patch.format],
                       shape=[patch.height, patch.width, patch.bands])
```

```python
image = pyvips.Image.new_from_file('../input/hubmap-kidney-segmentation/train/0486052bb.tiff')
patch = image.crop(10000, 10000, 2048, 2048)

np_img = np.ndarray(buffer=patch.write_to_memory(),
                   dtype=format_to_dtype[patch.format],
                   shape=[patch.height, patch.width, patch.bands])
plt.imshow(np_img)
```
