# Explore the geometry

We sample random crops from the image. Typically models take such crops for image segmentation. One difficulty is the limited amount of training examples. We suggest random sampling of crops. Another issue is the imbalance of pixels belonging to glomeruli in contrast to pixels that don't. Thus it may be advantageous to sample crops close to glomeruli. 

In the following we visualize the distribution of the number of glomeruli covered by a crop for the various techniques. 
We also visulize the crops (their distribution and size) on the original image to get a feel for the sampling at use. 

Note that each sample crop has a smaller box inside. For instance this can be useful if we classify crops by their glomeruli containment. In this instance we would decide only if at least one glomerulus intersects the inner box whereas the model sees the full box and in this way has some surrounding information available. 

```python
scope_size = 512 # crop in pxl
jiggle = 1 # std deviation from center (generated by pipeline)
overlap = 1/6 # overlap of crops
internal_prop = 3/4 # how much is in the core (decisive region) of a sampled box
```

```python
# uncomment what you want to try
grid_pipe = get_scopes_on_grid(img, scope_size, jiggle, overlap, internal_prop, seed=0)
rand_pipe = get_scopes_at_random(img, scope_size, internal_prop, seed=0)
glom_pipe = get_scopes_near_glomeruli(img, glomeruli, scope_size, jiggle, internal_prop, seed=0)
```

```python
nr_scopes = 200
plot_distribution_glomeruli_in_core(img, glomeruli, grid_pipe, simulation_runs=1000)
plot_scopes(img, glomeruli, grid_pipe, nr_scopes)
    
```

```python
nr_scopes = 200
plot_distribution_glomeruli_in_core(img, glomeruli, rand_pipe, simulation_runs=1000)
plot_scopes(img, glomeruli, rand_pipe, nr_scopes)
```

```python
nr_scopes = 200
plot_distribution_glomeruli_in_core(img, glomeruli, glom_pipe, simulation_runs=1000)
plot_scopes(img, glomeruli, glom_pipe, nr_scopes)
```
