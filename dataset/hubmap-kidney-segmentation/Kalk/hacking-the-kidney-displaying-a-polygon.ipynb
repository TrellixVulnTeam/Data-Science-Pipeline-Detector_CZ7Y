# Hacking the Kidney: Displaying a Polygon

This is a notebook meant for CV2 beginners. The goal of this is to show how to display a polygon on an image using CV2. 

### Imports

```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
import cv2
from osgeo import gdal 
import json

%matplotlib inline
```

### Load an Image

```python
dataset = gdal.Open(r'/kaggle/input/hubmap-kidney-segmentation/train/54f2eec69.tiff')
 
band1 = dataset.GetRasterBand(1) # Red channel 
band2 = dataset.GetRasterBand(2) # Green channel 
band3 = dataset.GetRasterBand(3) # Blue channel

b1 = band1.ReadAsArray() 
b2 = band2.ReadAsArray() 
b3 = band3.ReadAsArray() 

img = np.dstack((b1, b2, b3)) 
```

### Scale it so the CPU doesn't get overwhelmed

```python
img_small = cv2.resize(img, None,fx=0.5, fy=0.5)
```

### Get Polygon Dimensions

```python
def getJsonObj(jsonFile):
    with open(jsonFile, 'r') as myfile:
        data=myfile.read()

    # parse file
    obj = json.loads(data)
    
    return obj
```

```python
obj1 = getJsonObj('/kaggle/input/hubmap-kidney-segmentation/train/54f2eec69-anatomical-structure.json')
poly_coords = np.array(obj1[0]['geometry']['coordinates'][0], np.int32)
poly_coords2 = np.array(obj1[1]['geometry']['coordinates'][0], np.int32)

# scale the polygon to keep proportions (with scaled image)
temp_small = np.array(poly_coords/2, np.int32) 
temp_small2 = np.array(poly_coords2/2, np.int32)

#pts = temp.reshape((-1, 1, 2)) 
pts = temp_small.reshape((-1, 1, 2)) 
pts2 = temp_small2.reshape((-1, 1, 2)) 
```

### Draw Polygon

```python
img_poly = cv2.polylines(img_small, [pts], True, (0,255,0), thickness=50) 
```

### Display Image

```python
plt.imshow(img_poly) 
```

### Draw Another Polygon

```python
img_poly2 = cv2.polylines(img_poly, [pts2], True, (0,255,0), thickness=50) 
```

### Display Image (Updated)

```python
plt.imshow(img_poly2) 
```
