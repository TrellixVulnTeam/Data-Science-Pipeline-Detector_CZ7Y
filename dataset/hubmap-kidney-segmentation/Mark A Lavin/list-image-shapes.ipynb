```python
def verify_read(file_list):
    for file_name in file_list:
        image = tifffile.imread( file_name )
        print('img = {}, shape = {}, dtype = {}'.format(file_name,image.shape, image.dtype ) )
        del image
        gc.collect()
        
print( "before verify_read:", psutil.virtual_memory(), file = sys.stderr )

print( "train, verify_read:" )
file_list = glob.glob('/kaggle/input/hubmap-kidney-segmentation/train/*.tiff')
verify_read(file_list)

print( "test, verify_read:" )
file_list = glob.glob('/kaggle/input/hubmap-kidney-segmentation/test/*.tiff')
print( "len(file_list)", len(file_list) )
verify_read(file_list)

```
