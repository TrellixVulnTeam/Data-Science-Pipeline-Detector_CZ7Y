## Import packages

```python
import cv2
import json
import os
import csv
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw
```

## Load polygon mask

```python
json_filename = '../input/hubmap-kidney-segmentation/train/0486052bb.json'
read_file = open(json_filename, "r") 
data = json.load(read_file)
data[0]

polys = []
for index in range(data.__len__()):
    geom = np.array(data[index]['geometry']['coordinates'])
    polys.append(geom)
```

### The shape of the image, h * w

```python
shape = (25784, 34937)
```

## Reference code for converting polygon vector into pixel masks

### Method 1

```python
mask_1 = np.zeros(shape)
for i in range(len(polys)):
    cv2.fillPoly(mask_1, polys[i], i+1)

plt.imshow(mask_1)
```

### Method 2

```python
Image.MAX_IMAGE_PIXELS = None
img = Image.new('L', (shape[1], shape[0]), 0)  # (w, h)
for i in range(len(polys)):
    poly = polys[i]
    ImageDraw.Draw(img).polygon(tuple(map(tuple, poly[0])), outline=i + 1, fill=i + 1) 

mask_2 = np.array(img)
plt.imshow(mask_2)
```
