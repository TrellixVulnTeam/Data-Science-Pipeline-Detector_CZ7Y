# The Updated HuBMAP Glomeruli FTU Segmentation Dataset
#### by Leah Scherschel - Indiana University


The HuBMAP Tissue Mapping Center - Vanderbilt University (TMC-VU) team at the [BIOmolecular Multimodal Imaging Center (BIOMIC)](https://medschool.vanderbilt.edu/biomic/) has generously provided for use in this competition 10 additional PAS-stain whole slide images (WSIs), anatomical structure masks, and glomeruli segmentation masks which have never been published.

## What Has Changed

- The original 20 sets of images and masks (which were publicly available via the HuBMAP [Data Portal](https://portal.hubmapconsortium.org/)) have become the public train and public test datasets.
    - 15 train
    - 5 test
- The 10 new, unpublished sets are the new private test dataset.
- All 30 Glomeruli Segmentation masks (both new and original) have been thoroughly reviewed and improved. This fixed shifted masks and improved the quality of glomerulus annotations.
- A typo in the file HuBMAP-20-dataset_information.csv for image c68fe75ea.tiff's "width_pixels" field has been corrected.
- The timeline has been extended to provide competitors 2 months to work with the new data.
   
## What Has NOT Changed

- The pipeline used to generate the WSIs.
- The file formats.

Please refer to the prior [Dataset Details](https://www.kaggle.com/leahscherschel/dataset-details) notebook for more information on how the dataset was generated and file structure.
