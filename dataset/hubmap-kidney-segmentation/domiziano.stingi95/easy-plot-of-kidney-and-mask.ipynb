I tried to create the most easy and user friendly function to plot the picture with the relative mask

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt


import tifffile
```

```python
def get_mask(id_):
    train_csv = pd.read_csv('../input/hubmap-kidney-segmentation/train.csv')
    image = tifffile.imread('../input/hubmap-kidney-segmentation/train/{}.tiff'.format(id_))
    mask_cod  = train_csv.loc[train_csv['id'] == id_]['encoding'].values[0]
    
    mask = np.zeros((image.shape[0]*image.shape[1]), dtype=np.uint8)
    
    rle_mask = mask_cod.split()
    positions = map(int, rle_mask[::2])
    lengths = map(int, rle_mask[1::2])
    for pos, le in zip(positions, lengths):
        mask[pos-1:pos+le-1] = 1
   
    mask = mask.reshape((image.shape[1], image.shape[0]))

    return image, mask.T
```

```python
image, mask = get_mask('2f6ecfcdf')
```

```python
plt.figure(figsize=(15,15))
plt.imshow(image)
plt.imshow(mask,alpha=0.5)
```

I copied the next function from https://www.kaggle.com/ihelon/hubmap-exploratory-data-analysis/notebook
because it is definetly better than whatever I could ever produce 

```python
def plot_grid_image_with_mask(image, mask):
    plt.figure(figsize=(16, 16))
    n_cols = 4
    n_rows = 4
    col_start = 0
    col_w = 5000
    row_start = 5000
    row_w = 5000
    for i in range(n_cols):
        for j in range(n_rows):
            plt.subplot(n_cols, n_rows, n_rows * i + j + 1)
            sub_image = image[
                col_start + i * col_w : col_start + (i + 1) * col_w, 
                row_start + j * row_w : row_start + (j + 1) * row_w, 
                :
            ]
            sub_mask = mask[
                col_start + i * col_w : col_start + (i + 1) * col_w, 
                row_start + j * row_w : row_start + (j + 1) * row_w, 
            ]
            plt.imshow(sub_image)
            plt.imshow(sub_mask, alpha=0.5)
            plt.axis("off")
    plt.show()
```

```python
plot_grid_image_with_mask(image,mask)
```
