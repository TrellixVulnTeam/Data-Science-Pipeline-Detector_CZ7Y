# Introduction


This Notebook shows how you can extract with just few lines of code the coordinates for the anatomical structure and glomerii from the JSON data.

```python
import pandas as pd
import json
import os
import matplotlib.pyplot as plt
```

```python
def load_normalize_json_data(json_file):
    json_path = os.path.join("/kaggle/input/hubmap-kidney-segmentation/train",json_file)
    with open(json_path) as json_file:
        json_data = json.load(json_file)
    json_data_df = pd.json_normalize(json_data)
    return json_data_df
```

```python
anatomical = load_normalize_json_data("0486052bb-anatomical-structure.json")
anatomical.head()
```

```python
glomeri = load_normalize_json_data("0486052bb.json")
glomeri.head()
```

Now we can easily extract each set of geometry coordinates. For example, first formation geometry is defined by:
