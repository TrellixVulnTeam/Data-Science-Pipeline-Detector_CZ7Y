```python
import json
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw
Image.MAX_IMAGE_PIXELS = None
from skimage import io
```

# Convert JSON to Mask Image

```python
img_path = "../input/medical-labeling/test.tiff"
json_path = "../input/medical-labeling/test.json"
img = io.imread(img_path, plugin='pil')[:,:,:3]

# Conversion needs the dimension of the image to create
# a mask with the same size
y_max = img.shape[0]
x_max = img.shape[1]

# Open the mask file
read_file = open(json_path, "r") 
data = json.load(read_file)

# Add each polygon to list
polys = []
for index in range(data.__len__()):
    geom = np.array(data[index]['geometry']['coordinates'])
    polys.append(geom)


# "Draw" the polygons
msk = Image.new('L', (x_max, y_max), 0)  # (w, h)
for i in range(len(polys)):
    poly = polys[i]
    ImageDraw.Draw(msk).polygon(tuple(map(tuple, poly[0])), outline=1, fill=1) 

mask = np.array(msk)

# Visualize
fix, ax = plt.subplots(1,2, figsize=(15,10))
ax[0].imshow(img)
ax[1].imshow(mask)
```
