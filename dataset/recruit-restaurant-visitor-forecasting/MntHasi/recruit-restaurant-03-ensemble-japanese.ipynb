# データのインポート
下記コードで実行したF_submission_week_all (44).csvを読み込み<br>
[210516_recruit_pre(Japanese)](https://www.kaggle.com/mnthasi/210516-recruit-pre-japanese)<br>
[210516_recruit_week1_LightGBM(Japanese)](https://www.kaggle.com/mnthasi/210516-recruit-week1-lightgbm-japanese)

```python
# 真面目に機械学習してきたもの
sub_46 = pd.read_csv("../input/sub-46/F_submission_week_all (46).csv")
sub_46 = sub_46.rename(columns={'visitors': 'visitors46'})
sub_46
```

# 統計量から予測した値
2017年3,4月の店舗ごとの来店者数を用いて、testdataを予測した結果をインポート<br>
Import the result of predicting test data using the number of visitors per store in March and April 2017

```python
# 機械学習などしていないが最もスコアが高いもの
sub_19 = pd.read_csv("../input/submission-19/submission_19.csv")
sub_19 = sub_19.rename(columns={'visitors': 'visitors19'})
sub_19
```

# 結合
merge

```python
date_df_merge = pd.merge(sub_46, sub_19, on="id")
date_df_merge
```

# アンサンブルの割合を指定
Specify the ratio of the ensemble

```python
ratio19 = 0.35
ratio46 = 0.65
```

```python
date_df_merge["visitors"] = (date_df_merge["visitors19"]*ratio19 +
                             date_df_merge["visitors46"]*ratio46)
```

```python
date_df_merge
```

```python
date_df_merge = date_df_merge.drop(['visitors46', 'visitors19'], axis=1)
date_df_merge
```

# Export

```python
date_df_merge.to_csv("FU_submission_week_all.csv", index=False)
```
