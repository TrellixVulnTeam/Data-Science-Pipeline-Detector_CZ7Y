```python
# Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
import urllib.request

# DataFrame
wikiart = pd.read_csv('../input/-wikiart-all-images-120k-link/wikiart_scraped.csv')
```

# Looking for Mona Lisa ?

- Searching for 'mona lisa' in the *Artwork* column in order to find the original one

```python
df_mona_lisa = wikiart[wikiart['Artwork'].apply(lambda x: 'mona lisa' in x.lower())]
df_mona_lisa
```

- We can see the original is the one in the first row, with Leonardo da Vinci as the artist, let's plot it !

- Note: matplotlib requires RGB ordering whereas OpenCV (perversely) uses BGR

```python
url_response = urllib.request.urlopen(df_mona_lisa.iloc[0]['Link'])
img = cv2.imdecode(np.array(bytearray(url_response.read()), dtype=np.uint8), -1)
imageRGB = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)
plt.imshow(imageRGB);
```
