# Removing logos from pictures

I'll take the wild guess that Monet didn't put huge white logos in the corners of his work. Therefore, I've been trying to remove these huge logos in the pictures before using them in order to improve my results.

However, while doing research online I found that most of the codes already available need a mask of the logo to be removed. I don't really feel like manually making 500 masks on paint so I've been trying to find a way of removing (or at least blur) the logos.

This code is extremelly simple and does a pretty poor job at removing them as it changes other parts of the image besides the logo.

Next steps? Maybe train another GAN to remove logos and watermarks?

# 1 - Image with a logo

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('/kaggle/input/gan-getting-started/photo_jpg/03a2b7ce83.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 

plt.imshow(img)
plt.axis('off')
```

# Mask

```python
lower = np.array([100, 100, 100])
upper = np.array([255, 255, 225])
shapeMask = cv2.inRange(img, lower, upper)

shapeMask[40:215,40:215]=0

plt.imshow(shapeMask)
plt.axis('off')
```

# Image with the mask applied and smoothed

```python
dst = cv2.inpaint(img,shapeMask,3,cv2.INPAINT_TELEA)

plt.imshow(dst)
plt.axis('off')
```
