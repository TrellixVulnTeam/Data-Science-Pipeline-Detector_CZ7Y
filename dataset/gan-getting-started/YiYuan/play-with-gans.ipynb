```python
def load_image(image_path):
    max_dim=512

    img = tf.io.read_file(image_path)
    img = tf.image.decode_image(img, channels=3)
    img = tf.image.convert_image_dtype(img, tf.float32)

    shape = tf.cast(tf.shape(img)[:-1], tf.float32)
    long_dim = max(shape)
    scale = max_dim / long_dim
    new_shape = tf.cast(shape * scale, tf.int32)

    img = tf.image.resize(img, new_shape)
    img = img[tf.newaxis, :]

    return img
```

```python
def imshow(image, title=None):
    if(len(image.shape) > 3):
        image=np.squeeze(image, axis=0)
    plt.imshow(image)
    if(title):# if there's a title mention it
        plt.title(title)
```

```python

base_img=load_image('/kaggle/input/gan-getting-started/photo_jpg/bc89a879fd.jpg')
style_img=load_image('/kaggle/input/gan-getting-started/photo_jpg/f36fc36903.jpg')

plt.figure(figsize=(8,5))
plt.subplot(1, 2, 1)
imshow(base_img, 'Base Target Image')
plt.subplot(1, 2, 2)
imshow(style_img, 'Style image')
```
