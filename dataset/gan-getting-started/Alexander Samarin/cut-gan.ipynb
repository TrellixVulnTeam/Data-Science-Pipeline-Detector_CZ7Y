# Losses

Adversarial loss:

$$\mathcal{L}_{\operatorname{GAN}}(G,D,X,Y) = \mathbb{E}_{\mathbf{y}\sim Y}[\log D(y)] + \mathbb{E}_{\mathbf{x}\sim X}[\log(1-D(G(x)))] $$

The cross-entropy loss is calculated, representing the probability of the positive example being selected over negatives

$$\ell(v,v^+,v^-) = -\log \bigg[ \frac{\exp (v \cdot v^+ / \tau)}{\exp(v \cdot v^+ / \tau) + \sum_{n=1}^{N} \exp(v \cdot v_n^- / \tau)} \bigg]$$

PatchNCE loss
$$ \mathcal{L}_{\operatorname{PatchNCE}}(G, H, X) = \mathbb{E}_{\mathbf{x}\sim X} \sum_{l=1}^L \sum_{s=1}^{S_l} \ell (\hat{z}_l^s, z_l^s, z_l^{S \setminus s}) $$
