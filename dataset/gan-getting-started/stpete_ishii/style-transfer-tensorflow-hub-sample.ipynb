# Style Transfer TensorFlow Hub Sample
This notebook used TensorFlow Hub Pre-trained model.<br/>
https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/2

```python
import tensorflow_hub as hub
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
```

```python
model = hub.load('https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/2')
```

```python
def load_image(img_path):
    img = tf.io.read_file(img_path)
    img = tf.image.decode_image(img, channels = 3)
    img = tf.image.convert_image_dtype(img, tf.float32)
    img = img[tf.newaxis,:]
    return img
```

## Content image and Stylized image 

```python
content_image = load_image('../input/gan-getting-started/photo_jpg/009ddaed1f.jpg')
style_image = load_image('../input/gan-getting-started/monet_jpg/0e3b3292da.jpg')
```

```python
plt.figure(figsize = (10,5))

plt.subplot(1,2,1)
plt.title("content_image")
plt.imshow(np.squeeze(content_image))
plt.axis('off')

plt.subplot(1,2,2)
plt.title("style_image")
plt.imshow(np.squeeze(style_image))
plt.axis('off')
plt.show()
```

```python
stylized_image = model(tf.constant(content_image), tf.constant(style_image))[0]
```

```python
plt.figure(figsize = (5,5))

#plt.subplot(1,1,1)
plt.title("stylized_image")
plt.imshow(np.squeeze(stylized_image))
plt.axis('off')
plt.show()
```
