```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import tensorflow_hub as hub
import cv2
```

```python
tf.__version__
```

```python
model = hub.load('https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/2')
```

```python
model
```

```python
image_1="../input/gan-getting-started/monet_jpg/000c1e3bff.jpg"
image_2="../input/gan-getting-started/photo_jpg/00068bc07f.jpg"#
```

```python
def l(img):
    img=tf.io.read_file(img)
    img=tf.image.decode_image(img,channels=3)
    img=tf.image.convert_image_dtype(img,tf.float32)
    img=img[tf.newaxis,:]
    return img
```

```python
im1=l(image_1)
plt.imshow(np.squeeze(im1))
```

```python
im2=l(image_2)
plt.imshow(np.squeeze(im2))
```

```python
tfm=model(tf.constant(im1),tf.constant(im2))[0]
```

```python
plt.imshow(np.squeeze(tfm))
```

```python
photo_dir="../input/gan-getting-started/photo_jpg/"
monet_dir="../input/gan-getting-started/monet_jpg/"
```

```python
photo_path=[]
monet_path=[]
```

```python
for a,i in enumerate(os.listdir(photo_dir)):
    if a==10:
        break
    photo_path.append(os.path.join(photo_dir,i))
```

```python
photo_path
```

```python
for a,i in enumerate(os.listdir(monet_dir)):
    if a==10:
        break
    monet_path.append(os.path.join(monet_dir,i))
```

```python
monet_path
```

```python
im2=l(monet_path[0])
plt.imshow(np.squeeze(im2))
```

```python
im2=l(photo_path[0])
plt.imshow(np.squeeze(im2))
```

```python
sq=zip(photo_path,monet_path)
for (i,j) in sq:
    print(i,j)
    i=l(i)
    j=l(j)
    style=model(tf.constant(i),tf.constant(j))[0]
    plt.subplot(1,3,1)
    plt.title("content_image")
    plt.imshow(np.squeeze(i))
    plt.axis('off')

    plt.subplot(1,3,2)
    plt.title("style_image")
    plt.imshow(np.squeeze(j))
    plt.axis('off')

    plt.subplot(1,3,3)
    plt.title("stylized_image")
    plt.imshow(np.squeeze(style))
    plt.axis('off')
    plt.show()
```
