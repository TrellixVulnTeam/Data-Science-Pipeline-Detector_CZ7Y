````python
dfg=df.groupby('investment_id')
````

````python
from tqdm import tqdm
dfg=list(map(np.array,tqdm(dfg)))
````

````python
dfg[0][1].head()
````

````python
dfg=list(map(lambda x:x[1],dfg))
````

# Visualization

````python
targets=np.array(list(map(lambda x:np.array(x['target']),dfg)))
````

````python
targets=list(targets)
````

````python
print('maxlen of ',max(list(map(len,targets))))
````

# I am padding it to 750 because of lack of memory

````python
i=0

for _ in range(len(targets)):
    k=targets[i]
    if len(k)>=750:
        targets[i]=targets[i][:750]
        i+=1
    if len(k)<750:
        targets.pop(i)
````

````python
targets=np.array(targets)
targets.shape
````

````python
px.line(x=list(range(750)),y=np.mean(targets,axis=0))
````
