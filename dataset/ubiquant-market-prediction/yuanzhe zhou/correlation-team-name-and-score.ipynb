```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

lbs = []
lbs.append(pd.read_csv("../input/ubiquant-leaderboards/ubiquant-market-prediction-publicleaderboard.csv"))
lbs.append(pd.read_csv("../input/ubiquant-leaderboards/ubiquant-market-prediction-publicleaderboard_update1.csv"))
```

```python
from collections import defaultdict 
from tqdm.notebook import tqdm 

dict_name_score = defaultdict(list)

for name, score in lbs[0][['TeamName', 'Score']].values:
    words = name.lower().split(' ')
    
    for item in words:
        dict_name_score[item].append(score)

for name, score in lbs[1][['TeamName', 'Score']].values:
    words = name.lower().split(' ')
    
    for item in words:
        dict_name_score[item].append(score)        
```

```python
dict_name_score_avg = {}
dict_name_count = {}

for item  in dict_name_score:
    if len(dict_name_score[item]) < 3:
        continue
    dict_name_score_avg[item] = np.mean(dict_name_score[item])
    dict_name_count[item] = len(dict_name_score[item])
```

```python
df_score = pd.DataFrame({'name':[''], 'score':[0], 'count':[0] })
for index, item in enumerate(dict_name_score_avg.items()):
    df_score.loc[index] = [item[0], item[1], dict_name_count[item[0]]]
    
```
