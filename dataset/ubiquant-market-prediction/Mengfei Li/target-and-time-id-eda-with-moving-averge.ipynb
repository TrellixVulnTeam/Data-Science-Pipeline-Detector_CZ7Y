## Using the moving average, we can clearly see the time series distribution in the median values of target and all the features. Generally, the increase/ decrease trend is periodical.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
```

```python
train = pd.read_pickle('../input/ubiquant-market-prediction-half-precision-pickle/train.pkl')
train_tid_grouped = train.groupby('time_id')
```

```python
def lineplot(col="target", ma_steps=1):
    y = np.convolve(train_tid_grouped[col].median(), np.ones(ma_steps), mode='valid') / ma_steps
    x = np.arange(0,len(y))
    
    df = pd.DataFrame()
    df["time_id"] = x
    df[col] = y
    
    fig = px.line(df, x="time_id", y=col, title=col+': averaged in '+str(ma_steps)+' steps .')
    fig.show()
```

## Target:

```python
for ma_steps in [1, 10, 50]:
    lineplot("target", ma_steps=ma_steps)
```

## Feature f_0 as an Example:

```python
for ma_steps in [1, 10, 50]:
    lineplot("f_0", ma_steps=ma_steps)
```
