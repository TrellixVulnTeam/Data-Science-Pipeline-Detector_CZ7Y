```python
# let's see the bahavior of the target by time.
train['avg_target_per_time'] = train.groupby('time_id').target.transform('mean') 
```

```python
data = train[['time_id','avg_target_per_time']].set_index("time_id")
moving_average = data['avg_target_per_time'].rolling(window=1,
    center=True,      # puts the average at the center of the window 
).mean()              # compute the mean (could also do median, std, min, max, ...)

ax = data['avg_target_per_time'].plot(style=".", color="0.5",figsize=(24,5))
moving_average.plot(ax=ax, linewidth=3, title="avg_target_per time_id", legend=False);
```
