# Normalize Lag Features

This code normalize lag features by using StandardScaler.

Lag Datasete is here:
- f0~99: https://www.kaggle.com/takamichitoda/ump-lag-1-features-f0f99
- f100~199: https://www.kaggle.com/takamichitoda/ump-lag-1-features-f100f199
- f200~299: https://www.kaggle.com/takamichitoda/ump-lag-1-features-f200f299

```python
import gc
import pickle
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
%%time

f0_f99_lag_1 = np.load("/kaggle/input/ump-lag-1-features-f0f99/f0_f99_lag_1.npy")
f100_f199_lag_1 = np.load("/kaggle/input/ump-lag-1-features-f100f199/f100_f199_lag_1.npy")
f200_f299_lag_1 = np.load("/kaggle/input/ump-lag-1-features-f200f299/f200_f299_lag_1.npy")
```

```python
vecs = np.hstack([f0_f99_lag_1, f100_f199_lag_1, f200_f299_lag_1])
del f0_f99_lag_1, f100_f199_lag_1, f200_f299_lag_1
gc.collect()
```

```python
vecs.shape
```

```python
vecs = vecs.astype(np.float32)
gc.collect()
```

```python
np.save('lag_1_features', vecs)
```

```python
scaler = StandardScaler()
vecs = scaler.fit_transform(vecs)

pickle.dump(scaler, open("lag_1_std_scaler.pkl", "wb"))
np.save('lag_1_features_std_scaled', vecs)


gc.collect()
```
