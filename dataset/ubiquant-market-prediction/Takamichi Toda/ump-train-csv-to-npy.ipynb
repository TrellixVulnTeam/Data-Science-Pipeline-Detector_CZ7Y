# train.csv to npy

This notebook convert train.csv to npy file.

I use @colum2131's dataset to load the csv file faster.
https://www.kaggle.com/columbia2131/ump-train-picklefile

Thanks.

```python
import gc
import pickle
import numpy as np
import pandas as pd

from sklearn.preprocessing import StandardScaler, MinMaxScaler

import torch
import torch.nn as nn
from torch.nn import functional as F
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader

from tqdm.auto import tqdm
from matplotlib import pyplot as plt

import ubiquant
```

```python
class GCF:
    TRAIN_DATA = "/kaggle/input/ump-train-picklefile/train.pkl"
    FEAT_COLS = [f"f_{i}" for i in range(300)]
```

```python
train_df = pd.read_pickle(GCF.TRAIN_DATA)
```

```python
y = train_df["target"].values
time_id = train_df["time_id"].values
investment_id = train_df["investment_id"].values
del train_df['row_id'], train_df['time_id'], train_df['investment_id'], train_df['target']
gc.collect()

np.save('targets', y)
np.save('time_id', time_id)
np.save('investment_id', investment_id)
```

```python
X = train_df[GCF.FEAT_COLS].values

del train_df
gc.collect()

np.save('features', X)
```

```python
scaler = MinMaxScaler()
_X = scaler.fit_transform(X)

pickle.dump(scaler, open("mm_scaler.pkl", "wb"))
np.save('features_mm_scaled', _X)

del scaler, _X
gc.collect()
```

```python
scaler = StandardScaler()
_X = scaler.fit_transform(X)

pickle.dump(scaler, open("std_scaler.pkl", "wb"))
np.save('features_std_scaled', _X)

del scaler, _X
gc.collect()
```
