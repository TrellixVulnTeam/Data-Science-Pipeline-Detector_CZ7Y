# 3.verification

There is no point in letting it run too long, so limit it to 5 loops.<br>
If you want to verify memory leaks, etc., change this value as necessary.

```python
n_loop = 5
```

```python
train_dfs = pickle.load(open('../input/train-infer/train_df.pkl', 'rb'))
sample_prediction_dfs = pickle.load(open('../input/train-infer/train_submission.pkl', 'rb'))
```

```python
count = 0

for (test_df, sample_prediction_df) in zip(train_dfs, sample_prediction_dfs):
    
    print('*'*10)
    print(f'loop{count}')
    
    print('test_df')
    display(test_df)
    print('sample_prediction_df')
    display(sample_prediction_df)
    
    # Please copy and paste your code for inference below here
    feat_cols = [col for col in test_df.columns if col not in ['row_id', 'time_id', 'investment_id', 'target']]
    test_df = test_df[feat_cols]
    preds = [model.predict(test_df) for model in models]
    sample_prediction_df['target'] = np.mean(preds, axis=0)
    # Write anything other than env.predict above this.
    
    print('submission')
    display(sample_prediction_df) 
    
    count += 1
    
    if count > n_loop - 1:
        break
```
