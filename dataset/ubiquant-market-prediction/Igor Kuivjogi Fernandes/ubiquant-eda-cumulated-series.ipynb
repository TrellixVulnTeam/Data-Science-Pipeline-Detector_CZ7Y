```python
import os

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
plt.rcParams['figure.figsize'] = [8, 8]
plt.style.use('ggplot')
```

## EDA cumulated series

Plotting some cumulated values to check the relation between cumulated target and cumulated features.   

Work In Progress :) 

```python
%%time

train = pd.read_pickle('../input/ubiquant-market-prediction-half-precision-pickle/train.pkl')
train.head()
```

## Selecting some investments 

```python
train['investment_id'].value_counts().plot.hist(bins=50);
```

Most of the investments has ~1200 time ids

```python
some_investment_ids = train['investment_id'].value_counts().index[:10].tolist()
print(some_investment_ids)
```

```python
train_subset = train[train['investment_id'].isin(some_investment_ids)].reset_index(drop=True).copy()
train_subset
```

```python
def plot_cumulated_values(investment_id, features, ax):
    investment_series = train_subset[train_subset['investment_id'] == investment_id]
    for col in features:
        ax.plot(investment_series[col].cumsum(), label=col)
    ax.plot(investment_series['target'].cumsum(), linewidth=5, linestyle='dashed', label='target')
    ax.set_title(f'investiment_id = {investment_id}\nLength = {len(investment_series)}')
    ax.legend();
```

## Checking cumulated series

```python
fig, ax = plt.subplots(ncols=3, figsize=(20, 10))

plot_cumulated_values(
    investment_id=194,
    features=[f'f_{i}' for i in range(10)],
    ax=ax[0]
)

plot_cumulated_values(
    investment_id=509,
    features=[f'f_{i}' for i in range(10)],
    ax=ax[1]
)

plot_cumulated_values(
    investment_id=1062,
    features=[f'f_{i}' for i in range(10)],
    ax=ax[2]
)
```
