# Comparing the daily plots 

```python
train_df = pd.read_parquet("../input/ubiquant-parquet/train_low_mem.parquet")
time_range = int(train_df["time_id"].max())
time_range
```

```python
# Look like there is an upstream screener/filter to select some investment for this downstream task.
train_df.groupby(["time_id"])["investment_id"].count().plot()
```

```python
train_df.groupby(["time_id"])["target"].mean().plot()
```

```python
# SSE Composite Index
sse_df = pd.read_csv("../input/5yr-sse-composite-index-as-of-20220122/000001.SS.csv").tail(time_range)
print(sse_df.shape)
sse_df["Adj Close"].plot()
```

```python
# S&P Composite Index
sp_df = pd.read_csv("../input/sp-500-stocks/sp500_index.csv").tail(time_range).reset_index(drop=True)
sp_df["S&P500"].plot()
```

# P-correlation between the index and all the columns

```python
# SSE (More features are related to the SSE vs S&P)
for f in ["target"] + [f"f_{i}" for i in range(0, 300)]:
    corr = train_df.groupby("time_id")[f].mean().corr(sse_df["Adj Close"])
    if np.abs(corr) >= .3:
        print(f, corr)
```

```python
# S&P (Might be some features come from the US market?)
for f in ["target"] + [f"f_{i}" for i in range(0, 300)]:
    corr = train_df.groupby("time_id")[f].mean().corr(sp_df["S&P500"])
    if np.abs(corr) >= .3:
        print(f, corr)
```
