```python
import pandas as pd
import numpy as np
import os
from tqdm import tqdm
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_parquet('../input/ubiquant-10/ubiquant_sample.parquet')
data.head()
```

```python
train, test = train_test_split(data, test_size=0.2, random_state=42)
```

```python
target = 'target'
used = [x for x in train.columns if x != target]
```

```python
from autox.autox_competition.feature_selection import GRN_feature_selection
GRN_feature_selection = GRN_feature_selection()

column_definition = {
    "cat":['investment_id'],
    "num":[]
}
for i in range(300):
    column_definition['num'].append((f'f_{i}'))

GRN_feature_selection.fit(train[used], train[target], column_definition)
```

```python
GRN_feature_selection.feature2weight
```

```python
train_select = GRN_feature_selection.transform(train[used], top_k=20)
test_select = GRN_feature_selection.transform(test[used], top_k=20)
```
