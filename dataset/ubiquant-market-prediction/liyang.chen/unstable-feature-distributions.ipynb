```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train=pd.read_parquet('../input/ubiquanttrain/train.parquet')
```

```python
declist=[10,25,50,75,90]
features_deciles=dict()
stddict=dict()
meandict=dict()
for tim in train['time_id'].unique():
    print(tim)
    check=train[train['time_id']==tim]
    for i in range(300):
        if i not in features_deciles:
            features_deciles[i]=dict()
        for dec in declist:
            if dec not in  features_deciles[i]:
                 features_deciles[i][dec]=[]
            features_deciles[i][dec].append(np.percentile(check[f'f_{i}'],dec))
        if i not in stddict:
            stddict[i]=[]
        stddict[i].append(np.std(check[f'f_{i}']))
        
        if i not in meandict:
            meandict[i]=[]
        meandict[i].append(np.mean(check[f'f_{i}']))
```

```python
for i in range(300):
    fig = plt.figure(figsize=(10,5))
    for dec in declist:
        plt.plot(train['time_id'].unique(),features_deciles[i][dec])    
    #plt.plot(train['time_id'].unique(),np.array(stddict[i])*10,'--')
    plt.title(f'f_{i}    10%,25%,50%,75%,90% position value')
    figure, axes1 = plt.subplots(figsize=(10, 5))
    #axes.plot(train['time_id'].unique(),stddict[i],'--')
    axes1.plot(train['time_id'].unique(),meandict[i],label='mean value')
    axes1.tick_params(axis='y', colors='b')
    axes1.legend(loc=1)
    axes2 = axes1.twinx()
    axes2.spines['left'].set_color('b')
    axes2.plot(train['time_id'].unique(),stddict[i],'--',c='red',label='standard deviation')
    axes2.spines['right'].set_color('red')
    axes2.tick_params(axis='y', colors='red')
    axes2.legend(loc=9)
```
