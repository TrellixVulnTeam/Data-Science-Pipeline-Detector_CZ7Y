# 0. Import libraries

```python
import pandas as pd
import pickle
import lightgbm as lgb
from sklearn.model_selection import train_test_split
```

 # 1. Load data (pickle file)
 
 - The original data is too large to load, so we will load the Pickle file created in [this note](https://www.kaggle.com/shashimo/ubiquant-how-to-make-pickle-file).

```python
#Importing pickle files
with open('/kaggle/input/ubiquanttrainpicklefile/train.pickle', 'rb') as f:
    train = pickle.load(f)
```

# 3. LGBM

```python
train.drop(['row_id','time_id'], axis=1, inplace=True)
y=train.target
x=train.drop('target',axis=1)
del train
train_x,val_x,train_y,val_y=train_test_split(x,y,test_size=0.25,random_state=1) 
del y
del x
```

```python
train_x.head()
```

```python
model=lgb.LGBMRegressor(random_state=0)
model.fit(train_x,train_y,eval_set=[(val_x,val_y),(train_x,train_y)],verbose=50)
```

```python
lgb.plot_metric(model)
```

```python
import ubiquant
env = ubiquant.make_env()
iter_test = env.iter_test()
for (test_df, sample_prediction_df) in iter_test:
    test_df.drop(['row_id'], axis=1, inplace=True)
    pred=model.predict(test_df)
    sample_prediction_df['target'] = pred
    env.predict(sample_prediction_df)
```
