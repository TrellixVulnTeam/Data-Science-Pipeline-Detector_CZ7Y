```python
import pickle 
```

```python
loaded_model = pickle.load(open('../input/notebookb1a525f625/modeling', 'rb'))
```

```python
import ubiquant
env = ubiquant.make_env()   # initialize the environment
iter_test = env.iter_test()    # an iterator which loops over the test set and sample submission
for (test_df, sample_prediction_df) in iter_test:
    testi=test_df.drop(['row_id','investment_id'],axis=1)
    y_preds=loaded_model.predict(testi)
    sample_prediction_df['target'] = y_preds  # make your predictions here
    env.predict(sample_prediction_df)   # register your predictions
```
