**Using Parquet File from https://www.kaggle.com/robikscube/ubiquant-parquet?select=train_low_mem.parquet for reducing memory consumption**


```python

%%time
df = pd.read_parquet("../input/ubiquant-parquet/train_low_mem.parquet")


```

```python
df.head()
```

```python
df.shape
```

```python
print("Unique Time Id:", df.time_id.nunique())
```

```python
print("Unique Investment Id:", df.investment_id.nunique())
```

```python
N = 0
for i in df.columns:
    count = df[i].nunique()
    print(i, ':', count)
    N += 1
    if N > 3:
        break
```

```python
df.head()
```

```python
df_time = df.groupby('time_id').size()
```

```python
df_time.shape
```

```python
import plotly.express as px
```

```python
df.time_id.unique()
```

```python
plt.figure(figsize=(20, 8))
sns.set(style="ticks", font_scale = 1)
ax = sns.lineplot(data=df_time, x=df_time.index, y=df_time.values)
sns.despine(top=True, right=True, left=False, bottom=False)

ax.set_xlabel('time_id',fontsize = 14,weight = 'bold')
ax.set_ylabel('Count of row_id',fontsize = 14,weight = 'bold')
plt.title('Number of row_ids for Each time_id Entry',fontsize = 16,weight = 'bold');
plt.grid()
```

Here, between 300 and 550 time_id drastically dropped.
