# Ubiquant: The simplest baseline ever

The metric of this competition is just the correlation coefficient between the true and predicted target, not the squared or absolute errors. Therefore, the simplest prediction of the target could be the feature that correlates most with the target from train data (see [this notebook](https://www.kaggle.com/sytuannguyen/ubiquant-data-correlation)). Good luck!!!

```python
import ubiquant
env = ubiquant.make_env()   # initialize the environment
iter_test = env.iter_test()    # an iterator which loops over the test set and sample submission
for (test_df, sample_prediction_df) in iter_test:
    sample_prediction_df['target'] = -test_df['f_231']  # make your predictions here
    env.predict(sample_prediction_df)   # register your predictions
```
