```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import gc

from sklearn.linear_model import Ridge
from sklearn.ensemble import ExtraTreesRegressor as et
from catboost import CatBoostRegressor as cbr
from lightgbm import LGBMRegressor as lgb
from xgboost import XGBRegressor as xgb

import warnings
warnings.simplefilter('ignore')
```

```python
train = pd.read_pickle('../input/ump195gb/train.pkl')
train.head(3)
```

```python
X = train.drop(['row_id', 'time_id', 'target'], axis=1)
y = train['target']

del train
gc.collect()
```

```python
model = et(n_estimators=30)
model.fit(X,y)
```

```python
import ubiquant
env = ubiquant.make_env()   # initialize the environment
iter_test = env.iter_test()    # an iterator which loops over the test set and sample submission
for (test_df, sample_prediction_df) in iter_test:
    X_test = test_df.drop(['row_id'], axis=1)

    sample_prediction_df['target'] = model.predict(X_test) # make your predictions here
    env.predict(sample_prediction_df)   # register your predictions
```
