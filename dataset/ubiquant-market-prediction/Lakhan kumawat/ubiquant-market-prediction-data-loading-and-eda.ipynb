## choosing a subset

##### Since, the data is related with the time, we are choosing the subset skipping 10 rows everytime (we need a sequential data)

```python
samples=list(range(0,3141410,10))
```

```python
df_subset=df.iloc[samples]
```

```python
df_subset.describe()
```

```python
# choosiing the subset sequencially
for column in df_subset.columns[2:]:
    sns.histplot(df_subset[column])
    plt.show()
```

## standard normal distributions!!
## since, mean =0 and standard deviations=1 
## This subset is representing the main data set

```python
# check for outliers
for column in df_subset.columns[3:]:
    sns.boxplot(df_subset[column])
    plt.show()
```

### outliers detected!!!

```python
sns.heatmap(df_subset[['time_id','investment_id','target']].corr(),annot=True)
```

# correlation of target with columns ('f_0' to 'f_299')

```python
correlations=df_subset[unknown_columns].corrwith(df_subset['target'])
```

```python
plt.figure(figsize=(20,40))
plt.subplot(3,2,1)
sns.heatmap(pd.DataFrame(correlations[0:50],columns=['correlation']),annot=True)
plt.subplot(3,2,2)
sns.heatmap(pd.DataFrame(correlations[50:100],columns=['correlation']),annot=True)
```

```python
plt.figure(figsize=(20,40))
plt.subplot(3,2,3)
sns.heatmap(pd.DataFrame(correlations[100:150],columns=['correlation']),annot=True)
plt.subplot(3,2,4)
sns.heatmap(pd.DataFrame(correlations[150:200],columns=['correlation']),annot=True)
```
