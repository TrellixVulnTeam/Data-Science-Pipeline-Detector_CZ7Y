```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
import warnings
warnings.filterwarnings("ignore")
```

Using @slawekbiel Feather dataset: https://www.kaggle.com/slawekbiel/ubiquant-trainfeather-32-bit

```python
%%time
train_data = pd.read_feather('../input/ubiquant-trainfeather-32-bit/train32.feather')
```

```python
investment_id_ref = 2140
```

```python
train_data_time = train_data[train_data.investment_id == investment_id_ref]
```

```python
Target = train_data_time.target
```

```python
for i in range(300):
    feature_name = 'f_'+str(i)
    
    if train_data_time[feature_name].nunique() == 1:
        print('f_'+str(i)+'is constant for a given investment id')
```

```python
for i in range(300):
    feature_name = 'f_'+str(i)
    
    if train_data_time[feature_name].nunique() == 1:
        print('f_'+str(i)+'is constant for a given investment id')
    else:
        Feature = train_data_time[feature_name]
        color = (random.random(), random.random(), random.random())
        fig, axs = plt.subplots(ncols=2, figsize=(24,8))
        sns.lineplot(train_data_time.time_id,Feature, color=color, ax=axs[0]).set(title = feature_name+' v.s. strat id '+str(investment_id_ref))
        sns.regplot(x=Feature,y=Target,color=color, order = 2, line_kws={"color": 'black'}, ax=axs[1]).set(title = feature_name+' v.s. strat id '+str(investment_id_ref))
        plt.show()
```
