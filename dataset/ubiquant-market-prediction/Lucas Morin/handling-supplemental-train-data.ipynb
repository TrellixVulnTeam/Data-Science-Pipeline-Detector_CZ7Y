# Handling dtypes while reading

```python
path = '../input/ubiquant-market-prediction/supplemental_train.csv'

dtypes = {
    'row_id': 'str',
    'time_id': 'uint16',
    'investment_id': 'uint16',
    'target': 'float32',
}

for i in range(300):
    dtypes[f'f_{i}'] = 'float32'
    
sup_train = pd.read_csv(path,usecols=list(dtypes.keys()),dtype=dtypes, nrows=nrows)
```
