```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
import pickle
from sklearn.linear_model import LinearRegression
import ubiquant
from xgboost import XGBRegressor
```

```python
with open('../input/showmaker/model_lr.pickle', 'rb') as f:
    model = pickle.load(f)
with open('../input/showmaker/feature_names.pickle', 'rb') as f:
    feature_names = pickle.load(f)
```

```python
feature_names.remove('target')
```

```python
feature_names.remove('time_id')
```

```python
env = ubiquant.make_env()
iter_test = env.iter_test()
for (test_df, sample_prediction_df) in iter_test:
    X = test_df[feature_names]
    X['time_id'] = test_df['row_id'].str[:3].astype(int)
    Y = model.predict(X)
    sample_prediction_df['target'] = Y
    env.predict(sample_prediction_df)
```
