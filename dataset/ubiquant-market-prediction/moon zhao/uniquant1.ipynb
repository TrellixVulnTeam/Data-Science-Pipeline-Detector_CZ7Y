```python
import pickle
import pandas as pd
import numpy as np
import os
import ubiquant
env = ubiquant.make_env()  
iter_test = env.iter_test()
loaded_model = pickle.load(open("/kaggle/input/lightgbmpkl/lightgbm_model.pkl", "rb"))
```

```python
week_dict = {1:-0.059491,2:-0.071106,3:-0.041710,4:-0.066163,5:-0.040695,6:-0.026548,7:-0.097911}
```

```python

for (test_df, sample_prediction_df) in iter_test:
    test_df.drop(['row_id','investment_id'], axis=1, inplace=True)
    X_test = test_df.loc[:, test_df.columns.str.contains('f_')]
    X_test.fillna(-999)
    y_pred_f = loaded_model.predict(X_test)
    
    test_df['weekday'] =(test_df['time_id']-1000)%7+1
    y_pred_time = test_df['weekday'].apply(lambda x:week_dict[x])
    
    y_pred = y_pred_f*0.5+y_pred_time*0.5    
    sample_prediction_df['target'] = y_pred
    env.predict(sample_prediction_df) 
```
