# f_0~299 histogram

```python
import pandas as pd
```

```python
#https://www.kaggle.com/robikscube/fast-data-loading-and-low-mem-with-parquet-files#Reading-as-Parquet-Low-Memory-(Fast-&-Low-Mem-Use)
train = pd.read_parquet('../input/ubiquant-parquet/train_low_mem.parquet')
```

```python
columns = [f"f_{i}" for i in range(300)]
train.hist(column = columns, bins = 100, figsize = (30,30));
```
