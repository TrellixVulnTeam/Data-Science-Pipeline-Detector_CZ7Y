# 【EDA】f_0~299 KMeans Clustering

```python
#loading data
#https://www.kaggle.com/ilialar/ubiquant-eda-and-baseline
data_types_dict = {
    'time_id': 'int32',
    'investment_id': 'int16',
    "target": 'float16',
}

features = [f'f_{i}' for i in range(300)]

for f in features:
    data_types_dict[f] = 'float16'
    
target = 'target'
```

```python
import pandas as pd
train_df = pd.read_csv('../input/ubiquant-market-prediction/train.csv', 
                       usecols = data_types_dict.keys(),
                       dtype=data_types_dict, 
                       index_col = 0)
```

```python
#sampling
sample_df = train_df.sample(frac = 0.01)
sample_df
```

## f_0~299 KMeans Clustering

```python
from sklearn.cluster import KMeans

#transport sample_df
T = sample_df[features].T

#KMeans(4cluster)
km = KMeans(n_clusters=4, init='k-means++', n_init=10, max_iter=300, tol=1e-04, random_state=0)
label = km.fit_predict(T)
label
```

```python
label0 = T[label==0]
label1 = T[label==1]
label2 = T[label==2]
label3 = T[label==3]
print('label0',label0.index.values)
print('label1',label1.index.values)
print('label2',label2.index.values)
print('label3',label3.index.values)
```

I drew a histogram in [this](https://www.kaggle.com/mk1001/eda-f-0-299-histogram), and it seemed possible to divide the features into 4, so I set the number of clusters to 4 this time.
