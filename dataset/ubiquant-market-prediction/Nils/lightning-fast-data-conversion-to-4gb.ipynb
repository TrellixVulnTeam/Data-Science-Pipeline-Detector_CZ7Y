# Lighning fast data conversion with polars
Use `polars` to read the csv and use less memory consuming datatypes.
You can also convert the DataFrame to pandas with `to_pandas` and save as `feather` to get an even smaller file size.

```python
import numpy as np
import pandas as pd
import polars as pl
from pathlib import Path
```

```python
input_path = Path("/kaggle/input/ubiquant-market-prediction")
kaggle_home = Path("/kaggle/working")
[p for p in input_path.glob("*")]
```

```python
int32 = pl.Int32
float32 = pl.Float32
string = pl.Categorical

train_peek = pd.read_csv(input_path / "train.csv", nrows=2)
not_float_col_dtypes = {
    "row_id": string,
    "investment_id": int32,
    "time_id": int32,
    "target": float32,
}
float_cols = [c for c in train_peek.columns if "f_" in c]
float_col_dtypes = {
    col: float32 for col in float_cols
}
dtypes = {**float_col_dtypes, **not_float_col_dtypes}
```

```python
train = pl.read_csv(input_path / "train.csv", dtypes=dtypes)
train.to_parquet(kaggle_home / "train.parquet")
```
