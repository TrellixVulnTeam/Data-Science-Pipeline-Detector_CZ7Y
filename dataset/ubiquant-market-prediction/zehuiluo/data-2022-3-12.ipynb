```python
#check the datatype of int columns
def int_fit(d):
    dmax=max(d)
    dmin=min(d)
    if dmax >np.iinfo(np.int32).max or dmin <np.iinfo(np.int32).min:
        dtype='int64'
    elif dmax >np.iinfo(np.int16).max or dmin <np.iinfo(np.int16).min:
        dtype='int32'
    elif dmax >np.iinfo(np.int8).max or dmin <np.iinfo(np.int8).min:
        dtype='int16'
    else:
        dtype='int8'
    return dtype
    
    
with pd.read_csv('../input/ubiquant-market-prediction/train.csv',chunksize=10000) as reader:
    int_type1='int8'
    int_type2='int8'
    for chunk in reader:
        
        dtype1=int_fit(chunk['time_id'])
        dtype2=int_fit(chunk['investment_id'])
        
        if int(int_type1[3:])<int(dtype1[3:]):
            int_type1=dtype1 
        if int(int_type2[3:])<int(dtype2[3:]):
            int_type2=dtypeï¼’

print(int_type1,int_type2)
del reader
gc.collect()
```
