### TL;DR: there are 195 unseen investment_ids in the Public LB.

This notebook follows a similar approach to Lucas Morin's [Don't mind me just probing the LB](https://www.kaggle.com/lucasmorin/don-t-mind-me-just-probing-the-lb), but for counting the number of unseen unique `investment_id`s instead of counting the number of `time_id`s.    + V7: tests if len(unseen_set) <= 195. PASSES =>  190 < len(unseen_set) <= 195

   + V9: tests if len(unseen_set) <= 194. FAILS  =>  194 < len(unseen_set) <= 195
   + **=> len(unseen_set) = 195**

```python
import numpy as np
import pandas as pd
import ubiquant
```

```python
col_subset = ['investment_id'] 
train = pd.read_parquet('../input/ubiquant-parquet/train.parquet', columns=col_subset)
```

```python
env = ubiquant.make_env()   
iter_test = env.iter_test()
```

```python
train_ivs = set(train.investment_id.values)
max_train_iv = max(train_ivs)
missing_ivs = set(range(max(train_ivs) + 1)).difference(set(train_ivs))
```

```python
unseen_set = set([])
for (data_df, pred_df) in iter_test:
    data_ivs = data_df.investment_id.to_list()
    for i in data_ivs:
        if (i > max_train_iv) or (i in missing_ivs):
            unseen_set.add(i)
    if len(unseen_set) > 194:
        break
    env.predict(pred_df)
```
