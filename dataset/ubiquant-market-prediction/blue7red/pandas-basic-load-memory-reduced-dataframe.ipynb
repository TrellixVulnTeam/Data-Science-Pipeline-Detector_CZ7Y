```python
import pandas as pd 
import numpy as np
```

```python
TRAIN_PATH = "../input/ubiquant-market-prediction/train.csv"
REDUCED_TRAIN_PATH = "reduced_train.csv"
```

```python
%%time 

dtypes = {
    'row_id' : np.object,
    'time_id' : np.uint16,
    'investment_id' : np.uint16,
    'target' : np.float16,
}
for i in range(300):
    dtypes[f"f_{i}"] = np.float16

    
reduced_train = pd.DataFrame()

chunksize = 10 ** 6

for chunk in pd.read_csv(TRAIN_PATH, chunksize=chunksize, dtype=dtypes):
    reduced_train = pd.concat([reduced_train, chunk], ignore_index=True)
    
reduced_train.head()
```
