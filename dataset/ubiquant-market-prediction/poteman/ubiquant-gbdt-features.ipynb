```python
import pandas as pd
import numpy as np
import os
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('../input/ubiquant-market-prediction-small-train/small_train.csv')
```

```python
target = 'target'
used = [x for x in df.columns if x != target]

X_train, X_test, y_train, y_test = train_test_split(df[used], df[target], test_size=0.2, random_state=42)
```

```python
from autox.autox_competition.feature_engineer import FeatureGbdt
featureGbdt = FeatureGbdt()
featureGbdt.fit(X_train, y_train, objective= 'binary', num_of_features = 50)
```

```python
lgb_feature_train = featureGbdt.transform(X_train)
lgb_feature_test = featureGbdt.transform(X_test)
```
