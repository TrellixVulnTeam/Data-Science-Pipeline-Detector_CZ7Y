```python
%%time
df_train = pd.read_parquet('../input/ubiquant-parquet/train_low_mem.parquet')
```

```python
# 学習に使用するデータを設定
x = df_train.drop(['target','row_id','time_id'],axis=1)

x = x.drop(v[100:],axis=1)
y = df_train['target']

del df_train
#x.dtypes
gc.collect()  
```

```python
from sklearn.multioutput import MultiOutputRegressor

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)

data_eval = lgb.Dataset(x_test,y_test) 
data_train = lgb.Dataset(x_train,y_train)


# # del y_train,y_test
# # del x_train,x_test 

# # LightGBM parameters
# params = {
#         "objective": "regression",
#         "metric": "rmse",
#         "num_leaves": 30,
#         "max_depth": 6,
#         "feature_fraction": 0.8,
#         "subsample_freq": 1,
#         "bagging_fraction": 0.7,
#         "min_data_in_leaf": 50,
#         "learning_rate": 0.2,
#         "boosting": "gbdt",
#         "lambda_l1": 0.4,
#         "lambda_l2": 0.4,
#         "verbosity": -1,
#         "random_state": 4,
#         'num_iterations': 100,
#         'early_stopping_round': 20,
# }

# # モデルの学習
# model = lgb.train(params,
#                   data_train, # トレーニングデータの指定
#                   valid_sets=[data_train, data_eval], # 検証データの指定
#                   )
```

```python
def objective(trial):
    num_leaves = trial.suggest_int('num_leaves', 10, 100)    
    max_depth = trial.suggest_int('max_depth',2, 20)
    
    params = {
        "objective": "regression",
        "metric": "rmse",
        "num_leaves": num_leaves,
        "max_depth": max_depth,
        "feature_fraction": 0.8,
        "subsample_freq": 1,
        "bagging_fraction": 0.7,
        "min_data_in_leaf": 30,
        "learning_rate": 0.2,
        "boosting": "gbdt",
        "lambda_l1": 0.4,
        "lambda_l2": 0.4,
        "verbosity": -1,
        "random_state": 40,
        'num_iterations': 500,
        'early_stopping_round': 10,
    }

    # モデルの学習
    model = lgb.train(params,
                  data_train, # トレーニングデータの指定
                  valid_sets=[data_train, data_eval], # 検証データの指定
                  )
    
    y_tests=pd.DataFrame(y_test)
    y_tests['val']=model.predict(x_test)
    retu = y_tests.corr()["val"]["target"]

    return retu

#study = optuna.create_study(sampler=optuna.samplers.RandomSampler(seed=0))
study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=20)
study.best_params
```

```python
eval_param = study.best_params
```

```python
df_feature = pd.DataFrame(model.feature_importance(importance_type = 'gain'),
                           index=x.columns,columns=['importance'])
df_feature = df_feature.sort_values('importance', ascending=False)

plt.figure(figsize=(40, 12))
sns.barplot(x=x.columns, y='importance', data=df_feature)
plt.title(f'Head LightGBM Feature Importance')
plt.tight_layout()
plt.show()
```
