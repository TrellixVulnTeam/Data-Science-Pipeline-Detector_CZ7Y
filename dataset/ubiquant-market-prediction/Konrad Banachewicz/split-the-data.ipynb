```python
import numpy as np 
import pandas as pd
import os
```

```python
# Thanks to https://www.kaggle.com/danielkorth/ubiquant-trainfeather
train_df = pd.read_feather('../input/ubiquant-trainfeather/train.feather')
train_df.head()

# # store the part with time_id <= 800
train_df.loc[train_df.time_id <= 800].to_csv('xtrain800.csv', 
                                             index = False) # , compression="gzip")
idx = (train_df.time_id > 800) & (train_df.time_id <= 1000)
train_df.loc[idx].to_csv('xvalid800.csv',index = False) #, compression="gzip")
train_df.loc[idx][['row_id', 'target']].to_csv('xref800.csv', index = False) 

# cutoff at 1000
train_df.loc[train_df.time_id <= 1000].to_csv('xtrain1000.csv', 
                                              index = False) #, compression="gzip")

idx = (train_df.time_id > 1000) & (train_df.time_id <= 1300)
train_df.loc[idx].to_csv('xvalid1000.csv', index = False) 
train_df.loc[idx][['row_id', 'target']].to_csv('xref1000.csv', index = False) 
```
