## Data Loading 
* the training data size if 18 G, so don't try to load using read_csv, it will not work, 
* Instead try to use other methods to load big datasets https://www.kaggle.com/rohanrao/tutorial-on-reading-large-datasets 
* Or even use the existing pickle file, https://www.kaggle.com/columbia2131/ump-train-picklefile (3 GB)
* In order to further reduce the memory, use float16 instead of float32 for the features. This helps in saving another 1 GB (2 GB instead of ogirinal 18 GB) 

```python
!ls -lhtr /kaggle/input/ump-train-picklefile/train.pkl
```

```python
%%time 
df_train = pd.read_pickle("/kaggle/input/ump-train-picklefile/train.pkl")
```

```python
df_train
```

```python
#df['col'].astype(np.float16)
features= [f'f_{i}'for i in range(0,300)]
for ifeature in features:  ## this will take a few mins but need not repeat it, so its ok to spend 
    df_train[ifeature]=df_train[ifeature].astype(np.float16)
```
