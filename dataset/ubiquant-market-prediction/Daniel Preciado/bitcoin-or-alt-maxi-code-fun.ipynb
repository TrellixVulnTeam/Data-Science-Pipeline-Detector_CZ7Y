```python
%%time
import numpy as np
import pandas as pd
from sklearn import *
from scipy.stats import pearsonr as ps
import gc; gc.enable()

import ubiquant
env = ubiquant.make_env()

p = '/kaggle/input/ubiquant-market-prediction/'
train = pd.read_csv(p + 'train.csv', nrows=500_000, error_bad_lines=False)
train.fillna(0, inplace=True) #prevent errors
```

```python
%%time
cols = [c for c in train.columns if c not in ['row_id','time_id','target']]
model = linear_model.LinearRegression(n_jobs=-1)
model.fit(train[cols], train.target)
pred = model.predict(train[cols]).clip(0.000002, 0.999998)
print('score', ps(train.target.values, pred)[0])
del train
del pred
gc.collect()
```

```python
%%time
ei_test = env.iter_test()
for (test, sub) in ei_test:
    test.fillna(0, inplace=True) #prevent errors
    sub['target'] = model.predict(test[cols]).clip(0.000002, 0.999998)
    gc.collect()
    env.predict(sub)
```
