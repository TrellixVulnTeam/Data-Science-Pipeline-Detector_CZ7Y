```python
import numpy as np
import pandas as pd
import tensorflow as tf
import ubiquant
```

```python
best_model = tf.keras.models.load_model('../input/copy-of-pythonash-dnn-with-leakyrelu/pythonash_model.h5')  # load model file
env = ubiquant.make_env()   
iter_test = env.iter_test()    
```

```python
for (test_df, sample_prediction_df) in iter_test:
#     test_df = make_dataset(test_df)  # format input
    f_col = [f for f in test_df.columns if f not in ['row_id','time_id','investment_id']]
    test_df = test_df[f_col]

    predict_col = best_model.predict(test_df)  # model predict
    
    sample_prediction_df['target'] = predict_col
    env.predict(sample_prediction_df)
```
