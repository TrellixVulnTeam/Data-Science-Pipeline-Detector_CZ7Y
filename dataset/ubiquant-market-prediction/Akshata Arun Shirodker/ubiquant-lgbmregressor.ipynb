```python
# data is too large and the file crashes so I took data of 50000 to train and test the data

X_train,X_test,y_train,y_test=train_test_split(X[:50000],y[:50000],test_size=0.3,random_state=40)
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```
