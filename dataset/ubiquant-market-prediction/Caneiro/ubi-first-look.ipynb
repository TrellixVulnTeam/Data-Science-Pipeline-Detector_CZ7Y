# Data load
https://www.kaggle.com/robikscube/ubiquant-parquet  
With this parquet dataset the data loading is fast and less memory is used. Since we just gonna look to target variable right now we are not be importing all columns.

```python
col_subset = ['time_id','investment_id','target']
train = pd.read_parquet('../input/ubiquant-parquet/train_low_mem.parquet', columns=col_subset)
print(train.info())
print(train.shape)
train.head()
```

# Target ðŸ‘€

```python
train.target.hist()
```

```python
g = sns.relplot(data=train, x='time_id', y='target', kind='line', aspect=3)
```

# Investment id

```python
inv_ids = train.investment_id.unique()
print(inv_ids.shape[0])
print(inv_ids.max())
print(inv_ids.min())
```

```python
train.investment_id.value_counts().hist()
```

# Sample some investiments

```python
dfs = []
for iid in np.random.choice(inv_ids, 10, replace=False):
    df = pd.read_parquet('../input/ubiquant-parquet/investment_ids/{}.parquet'.format(iid))
    dfs.append(df)    

df = pd.concat(dfs).reset_index(drop=True)
```

```python
df.head()
```

```python
sns.relplot(data=df, x='time_id', y='target', col='investment_id', col_wrap=5, height=3)
```

Most asset don't have target data for all period. Let's investigate if same happens for other random selected variables.
