```python
from sklearn.model_selection import train_test_split
target = 'target'
used = [x for x in data.columns if x != target]

X_train, X_test, y_train, y_test = train_test_split(data[used], data[target], test_size=0.2, random_state=42)
```

```python
from autox.autox_competition.feature_engineer import FeatureGbdt
featureGbdt = FeatureGbdt()
featureGbdt.fit(X_train, y_train, objective= 'binary', num_of_features = 50)
```

```python
lgb_feature_train = featureGbdt.transform(X_train)
lgb_feature_test = featureGbdt.transform(X_test)
```

```python
lgb_feature_train.shape, lgb_feature_test.shape
```

```python
lgb_feature_train.head()
```

```python
from lightgbm import LGBMRegressor

model = LGBMRegressor()
model.fit(lgb_feature_train,y_train)
y_pred = model.predict(lgb_feature_test)
```

```python
from sklearn.metrics import mean_squared_error

print(mean_squared_error(y_test,y_pred))
```
