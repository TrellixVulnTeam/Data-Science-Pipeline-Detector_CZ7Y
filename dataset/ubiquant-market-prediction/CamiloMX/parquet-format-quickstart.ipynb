**Parquet format**

Apache Parquet is a free and open-source column-oriented data storage format.
It provides efficient data compression and encoding schemes with enhanced performance to handle complex data in bulk.

**File size comparison for Ubiquant Competition Dataset**
* CSV 17.2GB
* parquet 5.5GB (Using float 64)
* parquet 3.6GB (Using float 32)

**For using parquet format in your kaggle notebook you have to**
* add input data in your kaggle notebook and search for ubiquant-parquet

**For using parquet format in your pc you have to** 
* download the dataset in parquet format (https://www.kaggle.com/robikscube/ubiquant-parquet)
* and install pyarrow (pip install pyarrow)

```python
#%% Import Packages

#For using parquet format in your pc you have to install pyarrow
#pip install pyarrow

import pandas as pd

#%% Import Dataset
# Total rows = 3,141,410. Total time_ids = 3579
train = pd.read_parquet('../input/ubiquant-parquet/train_low_mem.parquet') # Import Dataset


#%% Select subsets of the dataset
train1 = train.loc[train.time_id <= 99] # Data from time_id 0 to 99 (226,894 rows)
train2 = train.loc[0:99999,:] # Data from row 0 to 99,999 (100,000 rows)


#%% Select X and y

y = train2.target  # Select y (the value we have to predict)

features = ['time_id','investment_id','f_0','f_1','f_2','f_3','f_4','f_5']
X = train2[features] # Select X (the features we want to use)


#%% Select specific values from  X and y
print(y[0])              # Print y in row 0
print(X.loc[0:10,"f_0"]) # First 10 values of f_0 (using column name)
print(X.iloc[0:10,3])    # First 10 values of f_1 (using column index)
```
