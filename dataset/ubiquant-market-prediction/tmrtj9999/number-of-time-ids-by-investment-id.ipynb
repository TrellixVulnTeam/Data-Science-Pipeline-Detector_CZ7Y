The set of investment_ids given is different for each time_id. Some of them appear frequently while others appear infrequently.

This time, I examined the number of times each investment_id appeared and did an EDA.

```python
import os
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')

import lightgbm as lgbm
from lightgbm import *
```

First, the data is read into a DataFrame.

```python
df = pd.read_parquet('../input/ubiquant-parquet/train_low_mem.parquet')
```

For each investment_id, the number of time_ids is tabulated and then graphed in descending order.

```python
from matplotlib import pyplot as plt
import pandas as pd
import matplotlib.pyplot as plt


time_count = df['investment_id'].value_counts()
time_count = time_count.reset_index()

time_ids = time_count['investment_id']

plt.plot(time_ids)
plt.show()
```

Then, in descending order, take the average value of target for each 100 stocks and graph it.

```python
inv_list = time_count['index'].values.tolist()
mean_list = []

for i in range(0, 36):
    df['flag'] = 0
    df.loc[df['investment_id'].isin(inv_list[100*i:100*(i+1)]), 'flag'] = 1

    df_2 = df[df['flag'] == 1]
    target_mean = df_2['target'].mean()

    mean_list.append(target_mean)
```

```python
from matplotlib import pyplot as plt
import pandas as pd
import matplotlib.pyplot as plt




plt.plot(mean_list)
plt.show()
```

Clearly, there are areas where the average value of TARGET is higher and areas where the value has dropped dramatically.

In addition, the standard deviation of TARGET for each 100 stocks in descending order is taken and graphed.

```python
from matplotlib import pyplot as plt
import pandas as pd
import matplotlib.pyplot as plt

inv_list = time_count['index'].values.tolist()
std_list = []

for i in range(0, 36):
    df['flag'] = 0
    df.loc[df['investment_id'].isin(inv_list[100*i:100*(i+1)]), 'flag'] = 1

    df_2 = df[df['flag'] == 1]
    target_std = df_2['target'].std()

    std_list.append(target_std)





plt.plot(std_list)
plt.show()
```

The further to the right (i.e., the fewer the number of time_ids investment_id), the larger the standard deviation of the target.
