```python
import pandas as pd
import pickle
from sklearn.linear_model import LinearRegression
import ubiquant
```

```python
with open('../input/linear-regression-example-training/model_lr.pickle', 'rb') as f:
    model = pickle.load(f)
with open('../input/linear-regression-example-training/feature_names.pickle', 'rb') as f:
    feature_names = pickle.load(f)
```

```python
env = ubiquant.make_env()
iter_test = env.iter_test()
for (test_df, sample_prediction_df) in iter_test:
    X = test_df[feature_names]
    Y = model.predict(X)
    sample_prediction_df['target'] = Y
    env.predict(sample_prediction_df)
```
