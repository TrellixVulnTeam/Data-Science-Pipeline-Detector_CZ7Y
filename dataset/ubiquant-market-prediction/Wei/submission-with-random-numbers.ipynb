```python
import os
import numpy as np
import pandas as pd
import ubiquant
env = ubiquant.make_env()   # initialize the environment
iter_test = env.iter_test()    # an iterator which loops over the test set and sample submission
```

```python
def predict():
    return np.random.rand() * np.random.rand() * (-1,1)[np.random.rand() < .5]
```

```python
for (test_df, sample_prediction_df) in iter_test:
    # randomly predict.
    sample_prediction_df['target'] = sample_prediction_df.apply(lambda x: predict(), axis=1)
    # return the prediction to API
    env.predict(sample_prediction_df)
    # probe the prediction
    print(sample_prediction_df)
```
