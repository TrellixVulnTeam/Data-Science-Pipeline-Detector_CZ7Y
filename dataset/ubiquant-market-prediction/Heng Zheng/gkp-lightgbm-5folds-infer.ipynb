```python
import warnings
warnings.simplefilter('ignore')

import gc

import numpy as np
import pandas as pd
pd.set_option('max_columns', None)
from tqdm.notebook import tqdm

import joblib
import lightgbm as lgb
```

```python
lgb_model_0 = joblib.load('../input/lightgbm5foldsgkp/lgb_0.pkl')
lgb_model_1 = joblib.load('../input/lightgbm5foldsgkp/lgb_1.pkl')
lgb_model_2 = joblib.load('../input/lightgbm5foldsgkp/lgb_2.pkl')
lgb_model_3 = joblib.load('../input/lightgbm5foldsgkp/lgb_3.pkl')
lgb_model_4 = joblib.load('../input/lightgbm5foldsgkp/lgb_4.pkl')
```

```python
import ubiquant
env = ubiquant.make_env()  
iter_test = env.iter_test()
```

```python
for (test_df, sample_prediction_df) in iter_test:
    test_df.drop(['row_id'], axis=1, inplace=True)
    pred_0 = lgb_model_0.predict(test_df)
    pred_1 = lgb_model_1.predict(test_df)
    pred_2 = lgb_model_2.predict(test_df)
    pred_3 = lgb_model_3.predict(test_df)
    pred_4 = lgb_model_4.predict(test_df)
    pred = (pred_0 + pred_1 + pred_2 + pred_3 + pred_4) / 5
    sample_prediction_df['target'] = pred
    env.predict(sample_prediction_df) 
```
