```python

import mytflib as tfl

rPATH = "../input/flower-classification-with-tpus/tfrecords-jpeg-512x512/train"
fPATH = [os.path.join(rPATH,ele) for ele in os.listdir(rPATH)]
tfrec_format = tfl.tfrec_format_generator({"image":"str","class":"int"})
ds = tf.data.TFRecordDataset(fPATH)
ds = ds.map(lambda example: tfl.parse_tfrecord_fn(example, tfrec_format))
ds = ds.shuffle(256)
```

## Displaying one image

```python
import matplotlib.pyplot as plt
next_item = next(iter(ds))
plt.imshow(
next_item['image']
)
```

## Displaying all 104 flowers!

```python
from tqdm import tqdm
# https://matplotlib.org/stable/gallery/subplots_axes_and_figures/subplots_demo.html
square = 16
FIGSIZE = (25,25)
figs, axs = plt.subplots(square, square, figsize= FIGSIZE)
for row in tqdm(range(square)):
    for col in range(square):
        next_item = next(iter(ds))
        axs[row,col].set_xticks([])
        axs[row,col].set_yticks([])
        axs[row,col].imshow(next_item['image'])
        axs[row,col].title.set_text("class: {}".format(next_item["class"]))
```
