```python
train_dataDir="../input/plant-seedlings-classification/train"
test_dataDir="../input/plant-seedlings-classification/."
```

```python
def copy(prefix="_"):
    datagen = ImageDataGenerator()
    for subfolder in os.listdir(train_dataDir):
    #     print(train_dataDir+"/"+ subfolder)
        if not os.path.exists("./{}".format(subfolder)):
            os.makedirs("./{}".format(subfolder))

        resized = datagen.flow_from_directory(train_dataDir,classes=[subfolder], target_size=(400, 400),
                                    save_to_dir="./{}".format(subfolder),
                                    color_mode="rgb", # Choose color mode
                                    class_mode=None,
                                    shuffle=True,
                                    save_prefix=prefix,
                                    save_format='png', # Formate
                                    batch_size=1)

        nums = len(os.listdir("./"+ subfolder))
        i=0
        while(i<len(resized) and nums+i<500):
                   resized.next()
                   i+=1
```

```python
def augument(prefix="_"):
    datagen = ImageDataGenerator(rotation_range=10, width_shift_range=0.1, height_shift_range=0.1,shear_range=0.15, zoom_range=0.1,channel_shift_range = 10, horizontal_flip=True)
    for subfolder in os.listdir(train_dataDir):
    #     print(train_dataDir+"/"+ subfolder)
        if not os.path.exists("./{}".format(subfolder)):
            os.makedirs("./{}".format(subfolder))

        resized = datagen.flow_from_directory(train_dataDir,classes=[subfolder], target_size=(400, 400),
                                    save_to_dir="./{}".format(subfolder),
                                    color_mode="rgb", # Choose color mode
                                    class_mode=None,
                                    shuffle=True,
                                    save_prefix=prefix,
                                    save_format='png', # Formate
                                    batch_size=1)
        
        nums = len(os.listdir("./"+ subfolder))
        i=0
        while(i<len(resized) and nums+i<500):
                   resized.next()
                   i+=1
```
