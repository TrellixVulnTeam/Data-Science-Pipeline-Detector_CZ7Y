### import necessary packages

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pydicom
from skimage.exposure import exposure, equalize_hist,equalize_adapthist
import cv2
from skimage.filters import unsharp_mask
print('packages imported')
```

### important function

#### Load dcm image

```python
def load_image(img_dcm):
    
    image = pydicom.dcmread(img_dcm)
    pixels = image.pixel_array

    min_pixel = np.min(pixels)
    max_pixel = np.max(pixels)

    if image.PhotometricInterpretation == "MONOCHROME1":
        pixels = max_pixel - pixels
    else:
        pixels = pixels

    return pixels


```

#### Filter function

```python
def apply_filter(f, img):
    if f == 'equalize_hist': 
        img = equalize_hist(img, nbins=256, mask= None)
    if f== 'equalize_adapthist':
        img = equalize_adapthist(img, kernel_size = None, clip_limit=0.01,nbins=256)
    if f == 'unsharp_mask':
        img = unsharp_mask(img, radius=5, amount=2)
    return img    
```

### Plot original and filtted image

```python

def plot_image(title, image, process_image):
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,12), sharex=True, sharey= True)
    ax = axes.ravel()
    ax[0].imshow(img, cmap=plt.cm.gray)
    ax[0].set_title('Original_image')
    ax[1].imshow(process_image, cmap=plt.cm.gray)
    ax[1].set_title(title)
    for a in ax:
        a.axis('off')
    fig.tight_layout()
    plt.show()
```

### Load an image and apply filters


```python


main_dir = '../input/siim-covid19-detection/train'
folder_names = os.listdir(main_dir)

for folder_name in folder_names:
    for img_folder in os.listdir(os.path.join(main_dir,folder_name)):
        for img_dir in os.listdir(main_dir+'/'+folder_name+'/'+img_folder):
            print(img_dir)
            img = load_image(main_dir+'/'+folder_name+'/'+img_folder+'/'+img_dir)
            processed_img = apply_filter('equalize_hist',img)
            plot_image('Image Equalize_hist', img, processed_img)
            
            processed_img = apply_filter('equalize_adapthist',img)
            plot_image('Image Equalize_adapthist', img, processed_img)
            
            processed_img = apply_filter('unsharp_mask',img)
            plot_image('Image unsharp_mask', img, processed_img)
    break
            
        
```
