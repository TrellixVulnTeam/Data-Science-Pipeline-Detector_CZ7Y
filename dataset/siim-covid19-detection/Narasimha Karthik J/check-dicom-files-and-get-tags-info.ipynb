## Read and Display Images

```python
import pydicom
from pydicom.data import get_testdata_files
from pydicom import dcmread
from fastai.medical.imaging import *
```

```python
def show_dcm_info(dataset):
#     print("Filename.........:", file_path)
#     print("Storage type.....:", dataset.SOPClassUID)
#     print()

    pat_name = dataset.PatientName
    display_name = pat_name.family_name + ", " + pat_name.given_name
    print("Patient's name......:", display_name)
    print("Patient id..........:", dataset.PatientID)
#     print("Patient's Age.......:", dataset.PatientAge)
    print("Patient's Sex.......:", dataset.PatientSex)
    print("Modality............:", dataset.Modality)
    print("Body Part Examined..:", dataset.BodyPartExamined)
    print("Study Instance UID.......:", dataset.StudyInstanceUID)
    
    if 'PixelData' in dataset:
        rows = int(dataset.Rows)
        cols = int(dataset.Columns)
        print("Image size.......: {rows:d} x {cols:d}, {size:d} bytes".format(
            rows=rows, cols=cols, size=len(dataset.PixelData)))
        if 'PixelSpacing' in dataset:
            print("Pixel spacing....:", dataset.PixelSpacing)
```

```python
def plot_pixel_array(dataset, figsize=(10,10)):
    plt.figure(figsize=figsize)
    plt.imshow(dataset.pixel_array, cmap=plt.cm.bone)
    plt.show()
```

```python
train_dcm_file_paths = get_dicom_files(train_path)
test_dcm_file_paths = get_dicom_files(test_path)
```

```python
print("Number of Train dcm files :", len(train_dcm_file_paths))
print("Number of Test dcm files :", len(test_dcm_file_paths))
```

```python
num_to_plot = 3
for i in range(num_to_plot):
    ds = train_dcm_file_paths[i].dcmread()
    show_dcm_info(ds)
    print()
    plot_pixel_array(ds)
    print("="*150)
    print()
```
