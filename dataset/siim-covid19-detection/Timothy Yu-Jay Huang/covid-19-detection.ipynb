# Import libraries

```python
import pydicom as dicom # dcm file
import matplotlib.pylab as plt # plot
from matplotlib import patches # bounding box
import numpy as np # linear algebra
import pandas as pd # csv
import os # directory, folder, file
import threading
print("pydicom, matplotlib.pylab, pydicom, os, pandas, and threading successfully imported")
```

# Define folder/file paths

```python
train_path = '../input/siim-covid19-detection/train'
test_path = '../input/siim-covid19-detection/test'
image_csv_path = '../input/siim-covid19-detection/train_image_level.csv'
study_csv_path = '../input/siim-covid19-detection/train_study_level.csv'
print("Successfully assign paths to variables")
```

# Read image and study leve csv files and convert them to Pandas' DataFrame

```python
study_df = pd.read_csv(study_csv_path)
image_df = pd.read_csv(image_csv_path)
```

# Display a patient's chest scan (utility function)

```python
def displayChestScan(id):
    ret = 0
    for dirname, dirs, filenames in os.walk(train_path):
        uid = dirname.split('/')[-2]
        if (uid == id):
            for filename in filenames:
                try:
                    image_path = os.path.join(dirname, filename)
                    ds = dicom.dcmread(image_path)
                    fig, img = plt.subplots(1, 1)
                    img.imshow(ds.pixel_array)
                    rect = patches.Rectangle((10, 10), 100, 100, 
                                                linewidth = 2,
                                                edgecolor = 'r',
                                                facecolor = 'none')
                    img.add_patch(rect)
                    plt.show()
                except Exception as e:
                    print('[Error][displayChestScan]', e)
                    plt.close()
                    ret = -1
            break
    
    return ret
```

# Display chest scan's bounding boxes and label (utility function)

```python
def displayChestScanInfo(id):
    for idx, row in image_df.iterrows():
        if (row['StudyInstanceUID'] == id):
            print(row)
            break
    
    return row
```

# Display chest scan by a user-given path (utility function)

```python
def displayChestScanByPath(path):
    ds = dicom.dcmread(path)
    plt.imshow(ds.pixel_array)
```

# Negative for Pneumonia

```python
error_log = {}
for _, row in study_df.iterrows():
    other_res = {row[key] for idx, key in enumerate(row.keys()) if (idx != 0 and idx != 1)} # 0: id, 1: Negative for Pneumonia
    id = row['id'].split('_')[0]
    neg_for_pnu = row['Negative for Pneumonia']
    if (neg_for_pnu and 1 not in other_res):
        ret = displayChestScan(id)
        displayChestScanInfo(id)
        
        if (ret is 0):
            break
```
