```python
import pandas as pd 
import numpy as np
label_csv=pd.read_csv('../input/siim-covid19-detection/train_image_level.csv')
print(label_csv.shape)
label_csv.head()
```

```python
cls_csv=pd.read_csv('../input/siim-covid19-detection/train_study_level.csv')
print(cls_csv.shape)
cls_csv.head(50)
```

```python
from sklearn.model_selection import GroupKFold
gkf=GroupKFold(n_splits=5)
label_csv['fold']=-1
for fold,(train_idx,val_idx) in enumerate(gkf.split(label_csv,groups=label_csv.StudyInstanceUID.tolist())):
    label_csv.loc[val_idx,'fold']=fold
label_csv.head(100)
label_csv.to_csv('lavel_fold.csv')
```

```python
cls_csv["study_id"] = cls_csv["id"].apply(lambda x: x.split("_")[0])
label_csv["image_id"]=label_csv["id"].apply(lambda x:x.split("_")[0])
```

```python
total = pd.merge(label_csv, cls_csv, 
                 left_on="StudyInstanceUID", right_on="study_id")
total.drop(["id_x", "StudyInstanceUID", "id_y"], axis=1, inplace=True)
total.head()
```
