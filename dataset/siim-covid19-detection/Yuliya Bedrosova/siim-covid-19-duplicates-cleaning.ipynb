Duplicates finding: https://www.kaggle.com/kwk100/siim-covid-19-duplicate-training-images

```python
import pandas as pd
```

```python
df_image = pd.read_csv('../input/siim-covid19-detection/train_image_level.csv')
df_study = pd.read_csv('../input/siim-covid19-detection/train_study_level.csv')
df_dupicates=pd.read_csv('../input/covidduplicates/duplicates.csv')
```

```python
df_dupicates.head()
```

```python
lduplicates=[]
for i in range(2,10):
    for image in df_dupicates.loc[:,str(i)]:
        if str(image)!='nan':
            lduplicates.append(image.replace('.dcm','_image'))
```

```python
image_duplicates=df_image[df_image.id.isin(lduplicates)]
```

```python
lduplicatess=image_duplicates.StudyInstanceUID+'_study'
```

```python
df_image.count()
```

```python
df_image=df_image.query("id not in @lduplicates")
```

```python
df_image.count()
```

```python
study_duplicates=df_study[df_study.id.isin(lduplicatess)]
```

```python
df_study.count()
```

```python
df_study=df_study.query("id not in @lduplicatess")
```

```python
df_study.count()
```

```python
image_duplicates.to_csv("train_image_level_deleted.csv",index=False)
study_duplicates.to_csv("train_study_level_deleted.csv",index=False)
df_image.to_csv("train_image_level_wndupls.csv",index=False)
df_study.to_csv("train_study_level_wndupls.csv",index=False)
```
