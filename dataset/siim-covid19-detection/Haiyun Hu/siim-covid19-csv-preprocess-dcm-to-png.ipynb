```python
import pandas as pd
import numpy as np
```

# CSV Preprocess

```python
study = pd.read_csv('../input/siim-covid19-detection/train_study_level.csv')
print(study.shape)
study.head()
```

```python
study.replace(regex=['_study'], value='',inplace=True)
study.head()
```

```python
study.rename(columns={"id":"StudyInstanceUID"}, inplace=True)
study.head()
```

```python
image = pd.read_csv('../input/siim-covid19-detection/train_image_level.csv')
print(image.shape)
image.head()
```

```python
image.replace(regex=['_image'], value='',inplace=True)
image.head()
```

```python
df = pd.merge(image,study,on="StudyInstanceUID")
df
```

```python
df.to_csv("siim-covid19-detection.csv",index=False)
```
