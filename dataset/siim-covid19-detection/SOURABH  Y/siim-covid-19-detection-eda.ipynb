```python
base_train_path='../input/siim-covid19-detection/train/'
```

```python
for i in range(len(image_level)):
    image_level['id'][i]=image_level['id'][i].split('_')[0]
```

```python
image_level.head()
```

```python
for i in range(1):
    path=glob(base_train_path+image_level['StudyInstanceUID'][i]+'/*/'+image_level['id'][i]+'.dcm')
    boxes=ast.literal_eval(image_level['boxes'][i])
    
    for p in path:
        ds=dcmread(p)
        fig, ax = plt.subplots(1, 1, figsize=(20, 4))
        for box in boxes:
            p = matplotlib.patches.Rectangle((box['x'], box['y']), box['width'], box['height'],
                                     ec='r', fc='none', lw=2.)
            ax.add_patch(p)
        ax.imshow(ds.pixel_array)
        plt.show()
    
```
