```python
import pydicom as dicom
import matplotlib.pylab as plt

ds = dicom.dcmread("../input/siim-covid19-detection/train/00086460a852/9e8302230c91/65761e66de9f.dcm")

plt.imshow(ds.pixel_array)
```

```python
import cv2   

def dicom2jpg(input_path, output_path):
    ds = dicom.dcmread(input_path)
    pixel_array_numpy = ds.pixel_array
    cv2.imwrite(output_path, pixel_array_numpy)
#Testing function
dicom2jpg("../input/siim-covid19-detection/train/00086460a852/9e8302230c91/65761e66de9f.dcm", "./65761e66de9f.png")
#Showing function output
%pylab inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
img = mpimg.imread('./65761e66de9f.png')
imgplot = plt.imshow(img)
plt.show()
```
