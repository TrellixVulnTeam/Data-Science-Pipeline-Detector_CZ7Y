```python
import numpy as np
import pandas as pd 
import os

ids1 = []
preds1 = []
ids2 = []
preds2 = []
for dirname, _, filenames in os.walk(f'../input/siim-covid19-detection/test'):
    for file in filenames:
        study_id = dirname.split('/')[-2] + '_study'
        if study_id not in ids1:
            ids1.append(study_id)
            preds1.append('negative 1 0 0 1 1 typical 1 0 0 1 1 indeterminate 1 0 0 1 1 atypical 1 0 0 1 1')
            
        image_id = file.replace('.dcm', '_image')
        ids2.append(image_id)
        preds2.append('OMG 1 0 0 1 1 Bug 1 0 0 1 1 ? 1 0 0 1 1')
out_csv = pd.DataFrame({
    'id': ids1,
    'PredictionString': preds1,
})
out_csv.to_csv('./submission_study.csv', index=False)

out_csv = pd.DataFrame({
    'id': ids2,
    'PredictionString': preds2,
})
out_csv.to_csv('./submission_image.csv', index=False)

img_csv = pd.read_csv('./submission_image.csv')
study_csv = pd.read_csv('./submission_study.csv')
final_csv = pd.concat([study_csv, img_csv])
final_csv.to_csv('./submission.csv', index=False)
print(final_csv)
```
