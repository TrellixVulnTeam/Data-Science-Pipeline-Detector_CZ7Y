```python
cls = np.array([0, 1, 2, 3])

for img_i in range(len(train_df)):
    
    img = train_df.index[img_i]
    res = ast.literal_eval(train_df.loc[img, "boxes"])
    assert type(res) is list
    
    
    label_inx = train_df.loc[img, "StudyInstanceUID"]
    arr_01_cols = label_df.loc[label_inx, :].values.astype(bool)
    
    this_cls = cls[arr_01_cols]
    
    f = open("./labels/"+img+".txt", "a")
    
    for bx_i in range(len(res)):

        
        f.write(str(0) +" "+ str((res[bx_i]["x"]+res[bx_i]["width"]/2)/train_df.loc[img, "dim1"]) + " " + str((res[bx_i]["y"]+res[bx_i]["height"]/2)/train_df.loc[img, "dim0"]) + " " + str(res[bx_i]["width"]/train_df.loc[img, "dim1"]) + " " + str(res[bx_i]["height"]/train_df.loc[img, "dim0"]))
        f.write("\n")
        
    f.close()
```
