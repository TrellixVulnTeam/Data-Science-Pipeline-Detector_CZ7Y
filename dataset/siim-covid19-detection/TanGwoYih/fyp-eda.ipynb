```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import glob
import cv2
import pydicom
```

# Study Level EDA

```python
train_study_df = pd.read_csv('../input/siim-covid19-detection/train_study_level.csv')
train_study_df['id'] = train_study_df.apply(lambda row: row.id.split('_')[0], axis=1)
train_study_df.columns = ['StudyInstanceUID', 'Negative for Pneumonia', 'Typical Appearance', 'Indeterminate Appearance', 'Atypical Appearance']
train_study_df.head(10)
```

```python
classes = ['Negative for Pneumonia', 'Typical Appearance', 'Indeterminate Appearance', 'Atypical Appearance']
```

```python
np.unique(train_study_df[classes].values, axis=0)
```

```python
classes_sum = train_study_df[classes].values.sum(axis=0)
classes_sum
```

```python
#  = np.array(images)
# label = np.array(label)
class_count_df = pd.DataFrame({'classes': classes, 'count': classes_sum}, columns=['classes', 'count'])
class_count_df
```

```python
fig = px.bar(class_count_df.sort_values('count'), 
             x=class_count_df["classes"].sort_index(), 
             y="count", 
             color=class_count_df["classes"], 
             opacity=0.5,
            title="<b>Number of Data per Labeled Category</b>",)

fig.update_layout(legend_title=None,
                  xaxis_title="",
                  yaxis_title="<b>Count</b>")

fig
```

# Image Level EDA
Image level has studies with more than one image

```python
train_image_df = pd.read_csv('../input/siim-covid19-detection/train_image_level.csv')
train_image_df
```

```python
train_image_df.StudyInstanceUID.value_counts()
```

```python
fig = px.histogram(train_image_df.StudyInstanceUID.value_counts(),
                   log_y=True,
                   opacity=0.5,
                   text_auto=True,
                   title="<b>Number of Images per Study</b>",)

fig.update_layout(showlegend=False,
                  xaxis_title="<b>Number of Images</b>",
                  yaxis_title="<b>Count</b>")

fig
```

```python
train_image_df['opacity'] = train_image_df.label.apply(lambda x: x.split()[0])
train_image_df
```

```python
fig = px.bar(train_image_df, 
             x=train_image_df['opacity'].unique(), 
             y=train_image_df['opacity'].value_counts(), 
             color=train_image_df['opacity'].unique(), 
             opacity=0.5,
            title="<b>Data with Opacity vs None</b>",)

fig.update_layout(legend_title=None,
                  xaxis_title="",
                  yaxis_title="<b>Count</b>")

fig
```

```python
# Merge image-level and study-level
train_study_df = train_study_df.merge(train_image_df, on='StudyInstanceUID',how="left")
train_study_df.head(10)
```
