```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
df=pd.read_csv('/kaggle/input/siim-covid19-detection/train_study_level.csv')
df
```

```python
x = df.iloc[:,1:4]
y = df.iloc[:,[4]]
```

```python
x.shape,y.shape
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
sns.countplot(df['Negative for Pneumonia'])

plt.subplot(1,2,2)
sns.countplot(df['Typical Appearance'])
```

```python
sns.pairplot(x,kind='scatter')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=57)
print('X_train shape = ',x_train.shape)
print('X_test shape = ',x_test.shape)
print('Y_train shape = ',y_train.shape)
print('Y_test shape = ',y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(criterion='gini',random_state=56)
dtc.fit(x_train,y_train)
```

```python
y_pred_dtc = dtc.predict(x_test)
y_pred_dtc
```

```python
print(confusion_matrix(y_test,y_pred_dtc))
print(accuracy_score(y_test,y_pred_dtc)*100)
```

```python
# Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=100,criterion='gini',random_state=55)
```

```python
rfc.fit(x_train,y_train)
y_pred_rfc = rfc.predict(x_test) 
```

```python
print(confusion_matrix(y_test,y_pred_rfc))
print(accuracy_score(y_test,y_pred_rfc)*100)
```

```python
# Logistic Regression 
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(random_state=55)
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred_lr = lr.predict(x_test)
```

```python
print(confusion_matrix(y_test,y_pred_lr))
print(accuracy_score(y_test,y_pred_lr)*100)
```

```python
# Support Vector Classification
from sklearn.svm import SVC
sv = SVC(kernel='rbf',random_state=55)
```

```python
sv.fit(x_train,y_train)
```

```python
y_pred_sv = sv.predict(x_test)
```

```python
print(confusion_matrix(y_test,y_pred_sv))
print(accuracy_score(y_test,y_pred_sv)*100)
```

```python
# K Nearest Neighbour Classifier
from sklearn.neighbors import KNeighborsClassifier
knc = KNeighborsClassifier(n_neighbors=7)
```

```python
knc.fit(x_train,y_train)
```

```python
y_pred_knc = knc.predict(x_test)
```

```python
print(confusion_matrix(y_test,y_pred_knc))
print(accuracy_score(y_test,y_pred_knc)*100)
```

```python
# Naive Bayes Classifier
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
```

```python
nb.fit(x_train,y_train)
```

```python
y_pred_nb = nb.predict(x_test)
```

```python
print(confusion_matrix(y_test,y_pred_nb))
print(accuracy_score(y_test,y_pred_nb)*100)
```
