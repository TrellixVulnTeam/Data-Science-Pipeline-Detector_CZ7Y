**Analysis of the submission samples for the [SIIM COVID-19 Detection](https://www.kaggle.com/c/siim-covid19-detection/overview) competition**.

**Conclusions**:
- The submission must include both image and study labels in the same file.
- The first two fields are related to studies (label, confidence), the next four fields to images (box dimensions).

````python
df = pd.read_csv(SAMPLES, index_col='id')
# assert df['id'].nunique() == len(df)
df.head()
````

# ANALYSIS

- There are 2,477 entries in total.
- There are no null values in the data.
- There is a single column: PredictionString

````python
df.info()
````

- There are two types of predictions:
    - **Study Level**:
        - The format for a given label's prediction would be:
            - a class ID
            - a confidence score
            - 0 0 1 1 is a one-pixel bounding box
        - Studies in the test set may contain more than one label.
        - Possible values: "negative", "typical", "indeterminate", "atypical".
        - For each study in the test set, you should predict at least one of the above labels.
    - **Image Level**:
        - Images in the test set may contain more than one object.
        - For each object in a given test image, you must predict:
            - a class ID of "opacity"
            - a confidence score
            - a bounding box (xmin ymin xmax ymax).
        - If you predict that there are NO objects in a given image, you should `predict none 1.0 0 0 1 1`.
            - where none is the class ID for "No finding", 1.0 is the confidence, and 0 0 1 1 is a one-pixel bounding box.
    
Source: https://www.kaggle.com/c/siim-covid19-detection/overview/evaluation
