**Analysis of the training studies for the [SIIM COVID-19 Detection](https://www.kaggle.com/c/siim-covid19-detection/overview) competition**.

**Conclusions**:
- There are no duplicates.
- The study labels are imbalanced.
- A study can only have one label (1).

# CONFS

```python
STUDIES = '/kaggle/input/siim-covid19-detection/train_study_level.csv'
```

# IMPORTS

```python
import pandas as pd
import plotly.express as px
```

# DATASETS

```python
df = pd.read_csv(STUDIES, index_col='id')
# assert df['id'].nunique() == len(df)
df.head()
```

# ANALYSIS

- IDs are unique (no duplicates).
- There is one label per case (1 value).

```python
df.info()
```

- <span style="color: red">**The dataset is imbalanced**</span>:
    - max. occurence: 47% (2,855 cases)
    - min. occurence: 7% (474 cases)

```python
cases = df.sum()
cases
```

```python
px.bar(cases, title='Number of studies per case')
```

```python
percents = df.sum() / len(df) * 100
percents
```

```python
px.pie(names=percents.index, values=percents.values)
```
