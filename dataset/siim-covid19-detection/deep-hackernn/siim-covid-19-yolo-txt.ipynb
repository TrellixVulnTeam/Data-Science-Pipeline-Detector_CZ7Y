```python
import pandas as pd
```

```python
df = pd.read_csv('../input/siim-covid19-detection/train_image_level.csv')
```

```python
df1 = pd.read_csv('../input/siim-covid19-resized-to-256px-jpg/meta.csv')
```

```python
df1 = df1[df1['split'] == 'train'].reset_index(drop=True)
```

```python
df1.head()
```

```python
df1.columns = ['id','dim0','dim1', 'split']
```

```python
df.head()
```

```python
df['id'] = df['id'].str[:-6]
```

```python
df = pd.merge(df, df1, on = 'id', how = 'left')
```

```python
df.head()
```

```python
for i in range(df.shape[0]):
    a = df.loc[i,'id']
    f = open(f'{a}.txt', 'w')
    b = df.loc[i, 'label'].split()
    b_len = int(len(b) / 6)
    if b[0] == 'none':
        f.close()
        continue   
    dim1 = df.loc[i,'dim1']
    dim0 = df.loc[i,'dim0']
    for j in range(b_len):
        x_mid = str((float(b[6*j+2]) + float(b[6*j+4])) / 2 / dim1)
        y_mid = str((float(b[6*j+3]) + float(b[6*j+5])) / 2 / dim0)
        w = str((float(b[6*j+4]) - float(b[6*j+2])) / dim1)
        h = str((float(b[6*j+5]) - float(b[6*j+3])) / dim0)
        f.write('0'+' '+x_mid +' '+y_mid+' '+w+' '+h+' ')
        f.write('\n')

    f.close()
```
