```python
filegroup='train'
path='../input/siim-covid19-detection'
dirName = f'{path}/{filegroup}/'
```

```python
# Get DICOM file list from train directory
items = get_dicom_files(dirName)
```

```python
df_trn = pd.DataFrame.from_dicoms(items,px_summ=False)
df_trn.to_csv(f'alldicom{filegroup}.csv')
df_trn.describe()
```
