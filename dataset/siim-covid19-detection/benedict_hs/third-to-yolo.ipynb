```python
import pandas as pd

df = pd.read_csv("../input/siim-covid19-detection/train_image_level.csv")
df_meta = pd.read_csv("../input/second/metadf.csv")
```

```python
df_meta.loc[:, "id"] += "_image"
df_meta
```

```python
df_meta.columns = ['id','dim0','dim1', 'split']
df = pd.merge(df, df_meta, on = "id", how = "left")
df
```

```python
for i in range(len(df)):
    
    id = df.loc[i, 'id']
    f = open(f"{id}.txt", 'w')
    x = df.loc[i, 'label'].split(" ")
    
    if x[0] == "none":
        continue

    dim0 = df.loc[i, "dim0"]
    dim1 = df.loc[i, "dim1"]
    label_len = int(len(x)/6)
    for j in range(label_len):
        x_mid = str((float(x[6*j+2]) + float(x[6*j+4]))/2/dim1)
        y_mid = str((float(x[6*j+3]) + float(x[6*j+5]))/2/dim0)
        width = str((float(x[6*j+4]) - float(x[6*j+2]))/dim1)
        height = str((float(x[6*j+5]) - float(x[6*j+3]))/dim0)
        f.write('0'+' '+x_mid+' '+y_mid+' '+width+' '+height+' ')
        f.write('\n')
    f.close()
```
