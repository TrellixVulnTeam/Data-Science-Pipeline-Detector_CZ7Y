# Basic Preprocessing
* Changing the data format to numeric type
* Dealing with missing values

```python
for label,content in df.items():
    if pd.api.types.is_string_dtype(content):
        print(label)
```

```python
for label, content in df.items():
    if pd.api.types.is_string_dtype(content):
        df[label] = content.astype('category').cat.as_ordered()
        df[label] = pd.Categorical(content).codes+1
```

```python
df.info()
```

```python
for label,content in df.items():
    if pd.api.types.is_numeric_dtype(content):
        if pd.isna(content).sum():
            print(label)
```

```python
for label,content in df.items():
    if pd.api.types.is_numeric_dtype(content):
        if pd.isna(content).sum():
            df[label] = content.fillna(content.median())
```

```python
df.info()
```

```python
df.isna().sum()
```

# Splitting the data as required 

```python
val_df = df[df.SaleYear == 2012]
train_df = df[df.SaleYear != 2012]
```

```python
x_train = train_df.drop('SalePrice',axis = 1)
y_train = train_df['SalePrice']
x_valid = val_df.drop('SalePrice',axis = 1)
y_valid = val_df['SalePrice']
```

```python
x_train.shape, y_train.shape, x_valid.shape, y_valid.shape
```

# Building an scoring Fuction

```python
def rmsle(y_test,y_preds):
    return np.sqrt(mean_squared_log_error(y_test,y_preds))

def scores(model):
    train_preds = model.predict(x_train)
    val_preds = model.predict(x_valid)
    scores = {"Training MAE": mean_absolute_error(y_train, train_preds),
              "Valid MAE": mean_absolute_error(y_valid, val_preds),
              "Training RMSLE": rmsle(y_train, train_preds),
              "Valid RMSLE": rmsle(y_valid, val_preds),
              "Training R^2": model.score(x_train, y_train),
              "Valid R^2": model.score(x_valid, y_valid)}
    return scores
```

# Model

```python
%%time
model = RandomForestRegressor(n_jobs = -1,
                              random_state = 42)
model.fit(x_train,y_train)
```

```python
scores(model)
```

# Hyperparameter tuning with RandomizedSearchCV

```python
%%time

grid = {"n_estimators": np.arange(10, 100, 10),
           "max_depth": [None, 3, 5, 10,15,20],
           "min_samples_split": np.arange(2, 20, 2),
           "min_samples_leaf": np.arange(1, 20, 2),
           "max_features": [0.5, 1, "sqrt", "auto"],
           "max_samples": [20000]}

model1 = RandomizedSearchCV(RandomForestRegressor(),
                              param_distributions= grid,
                              n_iter=40,
                              cv=5,
                              verbose=True)

model1.fit(x_train, y_train)
```

```python
scores(model1)
```

```python
model1.best_params_
```

# Training model with best Parameters

```python
%%time

tuned_model = RandomForestRegressor(n_estimators=60,
                                    min_samples_leaf=1,
                                    min_samples_split=12,
                                    max_features=0.5,
                                    n_jobs=-1)
tuned_model.fit(x_train, y_train)
```

```python
scores(tuned_model)
```
