```python
#importing spacy and LanguageDetector from spacy_langdetect
import spacy
from spacy_langdetect import LanguageDetector
```

## What's spaCy?
### spaCy is NLP Library which is considered as a successor to NLTK(Natural Language ToolKit - another most popular NLP library).
### spaCy's features include everything from tokenization, Named Entity Recognition to pretrained word vectors and support for languages as many as 59.
### Spacy's main website also claims to show spaCy as *the fastest Syntactic parser* in the world! Quiet an Achievement!

We will jump right into the Language Detection now

```python
def is_english(text: str) -> bool:
    """Simple Function to detect whether the language of the text is english or not.
    Returns a Boolean output for the same.Input the text column"""
    nlp = spacy.load('en')
    nlp.add_pipe(LanguageDetector(),name='language_detector',last=True)
    return nlp(text)._.language['language'] == 'en'
```

```python
#Applying the function
df['hypothesis'].apply(is_english)
```

So our simple function was able to detect the detect whether each text is on English Language or not. Same logic can be applied if you wanted to check for any other language and based out of the boolean output you can segregate the text data.
