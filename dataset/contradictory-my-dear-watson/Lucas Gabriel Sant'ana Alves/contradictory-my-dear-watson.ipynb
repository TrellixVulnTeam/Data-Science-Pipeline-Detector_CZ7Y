# Gráficos

```python
# Distribuição de classe em gráfico de pizza.

counts = df_train['label'].value_counts()
class_labels=['0', '1', '2']
counts_per_class=[counts[0], counts[1], counts[2]]
plt.figure(figsize=(5,10))
plt.pie(counts_per_class, labels=class_labels, autopct='%1.1f%%')
plt.title("Distribuição de classes")
plt.show()
```

```python
# Distribuição de classe em barra.

df_train.label.value_counts().plot(kind='bar', color=['#002045', '#d67822', '#f77f52'])
plt.xlabel("Classes")
plt.ylabel("Nº de ocorrências")
plt.title("Distribuição de classes")
plt.show()
```

```python
# Linguagens em gráfico de pizza.

labels, frequencies = np.unique(df_train.language.values, return_counts=True)
plt.figure(figsize=(10,10))
plt.pie(frequencies, labels=labels, autopct='%1.1f%%')
plt.show()
```

```python
# Linguagens em formato de barras.

plt.figure(figsize = (10,5))
sns.countplot(x='lang_abv', data=df_train)
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (15,5))
grafico = sns.countplot(df_train['language'], hue=df_train['label'])
grafico.set_title('Distribuição de Linguagens')
plt.show()
```
