```python
# Read train data
train = pd.read_csv('/kaggle/input/demand-forecasting-kernels-only/train.csv')

# Look at the shape of the data
print('Train shape:', train.shape)

# Look at the head() of the data
print(train.head())
```

```python
# Create a Random Forest object
rf = RandomForestRegressor()

# Train a model
rf.fit(X=train[['store', 'item']], y=train['sales'])
```

```python
# Read test and sample submission data
test = pd.read_csv('/kaggle/input/demand-forecasting-kernels-only/test.csv')
sample_submission = pd.read_csv('/kaggle/input/demand-forecasting-kernels-only/sample_submission.csv')

# Show the head() of the sample_submission
print(sample_submission.head())
```

```python
# Show the head() of the sample_submission
print(sample_submission.head())

# Get predictions for the test set
test['sales'] = rf.predict(test[['store', 'item']])

```

```python
# Write test predictions using the sample_submission format
# ./kaggle_submission.csv
test[['id', 'sales']].to_csv('./submission.csv', index=False)
```
