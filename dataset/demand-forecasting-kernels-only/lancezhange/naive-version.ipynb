```python
data_home = '/kaggle/input/demand-forecasting-kernels-only/'


train = pd.read_csv(f"{data_home}/train.csv", header=0, sep=',')
```

```python
train.head()
```

```python
train['date'] = pd.to_datetime(train['date'])

train['weekday']= train['date'].dt.weekday
train['month']  = train['date'].dt.month
```

```python
test = pd.read_csv(f"{data_home}/test.csv", header=0, sep=',')
```

```python

train = train.loc[train.date.dt.year==2017]


def naive(to_predict_day, to_predict_store, to_predict_item):
    
    to_predict_day = pd.to_datetime(pd.Series([to_predict_day]))
    
    wkday = to_predict_day.dt.weekday[0]
    
    month = to_predict_day.dt.month[0]
    
    
    
    
    history = train.loc[(train.weekday == wkday) & (train.month==month) &(train.store==to_predict_store) & (train.item==to_predict_item)]
    
    predicted = int(history.sales.mean())
    
    return predicted

    
    
```

```python
from tqdm import tqdm_notebook, _tqdm_notebook
_tqdm_notebook.tqdm_notebook.pandas()


test['preds'] = test.progress_apply(lambda x: naive(x['date'], x['store'], x['item']), axis=1)
```

```python
test[['id','preds']].rename(columns={'preds': 'sales'}).to_csv("submission.csv",
                                                       index=False,sep=',')
```
