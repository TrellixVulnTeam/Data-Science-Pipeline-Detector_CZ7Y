```python
#Importing data
train = pd.read_csv('/kaggle/input/demand-forecasting-kernels-only/train.csv')
train.shape
```

```python
train.info()
```

```python
#Converting date feature to date type object
train['date'] = pd.to_datetime(train['date'])
```

```python
#Setting date as index comlumn
train.set_index('date', inplace=True)
```

```python
train.head()
```

```python
train['store'] = train['store'].astype(str)
train['item'] = train['item'].astype(str)
```

```python
train.info()
```

```python
store_dummy = pd.get_dummies(train['store'])
store_dummy.shape
```

```python
item_dummy = pd.get_dummies(train['item'])
item_dummy.shape
```

```python
train = pd.concat([train,store_dummy], axis=1)
```

```python
train = pd.concat([train,item_dummy], axis=1)
train.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
X = train.drop('sales', axis=1)
y= train['sales']

X_train,X_test, y_train, y_test = train_test_split(X,y,test_size=0.01)
```

```python
#r_reg = RandomForestRegressor(n_estimators=3000, max_depth=8, max_features=3)
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
sgd = SGDRegressor(alpha=0.001)
sgd.fit(X_train_scaled,y_train)
```

```python
sgd.score(X_test_scaled,y_test)
```

```python
test = pd.read_csv('/kaggle/input/demand-forecasting-kernels-only/test.csv')
```

```python
test.head()
```

```python
test_new = test.drop('id', axis =1)
```

```python
test_new['date'] = pd.to_datetime(test['date'])
```

```python
test_new.set_index('date', inplace = True)
test_new
```

```python
st_dummy = pd.get_dummies(test_new['store'])
it_dummy = pd.get_dummies(test_new['item'])
```

```python
test_new = pd.concat([test_new,st_dummy],axis=1)
test_new = pd.concat([test_new,it_dummy],axis=1)
test_new.shape
```

```python
test_new_scaled = scaler.transform(test_new)
```

```python
y_preds=sgd.predict(test_new_scaled)
```
