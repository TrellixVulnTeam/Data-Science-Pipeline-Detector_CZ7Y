```python
import pandas as pd

train = pd.read_csv("/kaggle/input/demand-forecasting-kernels-only/train.csv")
test = pd.read_csv("/kaggle/input/demand-forecasting-kernels-only/test.csv")
```

```python
from catboost import CatBoost

model = CatBoost({"cat_features": ["store", "item"], "verbose": False})
model.fit(train.drop(columns=["date", "sales"]), train["sales"])
test_preds = model.predict(test.drop(columns=["id", "date"]))
```

```python
submission = pd.DataFrame({"id": test["id"], "sales": test_preds})
submission.to_csv("/kaggle/working/submission.csv", index=False)
```
