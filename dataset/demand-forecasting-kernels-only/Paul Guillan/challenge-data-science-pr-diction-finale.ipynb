```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from fbprophet import Prophet
```

Suite à l'échec sur la méthode utilisant un réseau de neurones LSTM, et à l'imprécision manifeste de la méthode de Holt-Winters pour des prédictions sur des durées aussi longues, on va donc utiliser la librairie Prophet pour prédire la demande sur trois mois.

```python
df_prophet_train = pd.read_csv("../input/demand-forecasting-kernels-only/train.csv")
df_prophet_train['date'] = pd.to_datetime(df_prophet_train['date'])
```

```python
df_prophet_train=df_prophet_train.drop(['store'],axis=1)
df_prophet_train=df_prophet_train.drop(['item'],axis=1)
df_prophet_train=df_prophet_train.head(5*365+1)
df_prophet_train.columns = ['ds','y']
df_prophet_train.head()
```

On met en forme le dataframe pour stocker et afficher la demande que l'on va prédire :

```python
df_prophet_predict = pd.read_csv("../input/demand-forecasting-kernels-only/test.csv")
df_prophet_predict['date'] = pd.to_datetime(df_prophet_predict['date'])
```

```python
df_prophet_predict=df_prophet_predict.drop(['store'],axis=1)
df_prophet_predict=df_prophet_predict.drop(['item'],axis=1)
df_prophet_predict=df_prophet_predict.drop(['id'],axis=1)
df_prophet_predict=df_prophet_predict.head(92)
df_prophet_predict=df_prophet_predict.assign(prediction=0)
df_prophet_predict.columns = ['ds','y']
df_prophet_predict.head()
```

```python
prophet = Prophet()
prophet.fit(df_prophet_train)
prediction = prophet.predict(df_prophet_predict)
```

```python
plt.figure(figsize=(12,7))
plt.plot(df_prophet_train['ds'] , df_prophet_train['y'], color = 'blue')
plt.plot(prediction['ds'] , prediction['yhat'], color = 'orange', alpha = 0.8)
plt.show()
```

```python
plt.figure(figsize=(12,7))
plt.plot(df_prophet_train['ds'] , df_prophet_train['y'], color = 'blue')
plt.plot(prediction['ds'] , prediction['yhat_upper'], color = 'red', alpha = 0.5)
plt.plot(prediction['ds'] , prediction['yhat_lower'], color = 'green', alpha = 0.5)
plt.show()
```
