```python
def visualize_a_day(self):
    for date in self['date'].unique():
        data1 = self[self['date'] == date]
        for num_store in data1['store'].unique():
            data2 = data1[data1['store'] == num_store]
            date = date.astype('datetime64[D]')
            fig = plt.figure(figsize=(10,8))
            ax = fig.add_subplot(111)
            ax.bar(data2['item'], data2['sales'], color='lightblue', label=date)
            ax.set_xlabel('Items_label', fontsize=18)
            ax.set_ylabel('Solded', fontsize=18)
            ax.set_title('Date:{}, Shop:{}'.format(date, num_store), fontsize=22)
            ax.grid(axis='y')
            ax.legend(fontsize=20)
            plt.show()
```

```python
visualize_a_day(tr_data)
```

```python
def make_constant(self):
    df = pd.DataFrame()
    for num_store in self['store'].unique():
        data1 = self[self['store'] == 1]
        for num_item in data1['item'].unique():
            data2 = data1[data1['item'] == num_item]
            for i in range(5):
                data2['sales_Lag{}'.format(i+1)] = data2['sales'].shift(i+1)   #shift関数でラグを再現
            df = df.append(data2)
    df = df.dropna(how='any', axis=0)
    df = df.reset_index(drop=True)
    return df
```
