```python
model = auto_timeseries(score_type='rmse',forecast_period=10,
                time_interval='D',
                non_seasonal_pdq=None, seasonality=False, seasonal_period=1,
                model_type=['ML'],dask_xgboost_flag=False, num_boost_rounds=1000,
                verbose=2)
```

```python
ts_column = 'date'
target = 'sales'
```

```python
model.fit(train, ts_column,target)
```

```python
predictions = model.predict( testdata = test, model = 'ML' )
predictions
```
