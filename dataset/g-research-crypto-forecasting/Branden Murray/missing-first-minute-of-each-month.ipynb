```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

data = pd.read_csv("../input/g-research-crypto-forecasting/train.csv")
```

Convert the timestamp to a more readable format.

```python
data['Time'] = pd.to_datetime(data['timestamp'], unit='s')
```

```python
print(data[(data.Time>='2018-01-01 23:59:00') & (data.Time<='2018-01-02 00:01:00')]['Time'].dt.time.unique())
print(data[(data.Time>='2019-03-15 23:59:00') & (data.Time<='2019-03-16 00:01:00')]['Time'].dt.time.unique())
print(data[(data.Time>='2020-09-23 23:59:00') & (data.Time<='2020-09-24 00:01:00')]['Time'].dt.time.unique())
print(data[(data.Time>='2021-07-08 23:59:00') & (data.Time<='2021-07-09 00:01:00')]['Time'].dt.time.unique())
```

However, as we transition from one month to the next, we no longer have the first minute of the month -- 00:00.

```python
print(data[(data.Time>='2018-01-31 23:59:00') & (data.Time<='2018-02-01 00:01:00')]['Time'].dt.time.unique())
print(data[(data.Time>='2019-03-31 23:59:00') & (data.Time<='2019-04-01 00:01:00')]['Time'].dt.time.unique())
print(data[(data.Time>='2020-09-30 23:59:00') & (data.Time<='2020-10-01 00:01:00')]['Time'].dt.time.unique())
print(data[(data.Time>='2021-07-31 23:59:00') & (data.Time<='2021-08-01 00:01:00')]['Time'].dt.time.unique())
```

Normally if the data for (t+1) or (t+16) is missing, then Target will be NaN. For example, for Asset_ID 11 there is no Target for 2018-01-01 00:01:00 because data is missing for 2018-01-01 00:17:00.

Also note that at 00:14:00 the Target is NaN because there is no data for the period after it, 00:15:00.

```python
data[(data.Asset_ID==11)][:40]
```

When the first minute of the month is missing we should expect there to be a missing Target 1 minute and 16 minutes prior, however there are no missing targets at all!

```python
data[(data.Time>='2018-01-31 23:40:00') & (data.Time<='2018-02-01 00:01:00') & (data.Asset_ID==1)][-30:]
```

So how is the target being calculated when the first minute of the month is missing? Is it using the Close price from 00:01:00? Or is is being calculated based on the Close at 00:00:00 -- data which we cannot see?
