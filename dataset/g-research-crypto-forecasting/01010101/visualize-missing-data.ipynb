```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
mpl.rcParams['axes.grid'] = False
```

```python
asset_df = pd.read_csv('../input/g-research-crypto-forecasting/asset_details.csv')
```

```python
train_df = pd.read_csv('../input/g-research-crypto-forecasting/train.csv')
# train_df.info()
train_df.head()
```

```python
sorted_train_df = train_df.sort_values(['Asset_ID','timestamp'])
# sorted_train_df.head()
```

```python
dropped_sorted_train_df = sorted_train_df.drop(columns=['timestamp','Count','Open','High','Low','Close','Volume','VWAP'])
# dropped_sorted_train_df.head()
```

```python
missing_ratio_lst = []
asset_name_lst = []

for asset_id in range(13):
    target_df = dropped_sorted_train_df[dropped_sorted_train_df['Asset_ID']==asset_id]
    n_data = len(target_df)
    n_miss = target_df['Target'].isnull().sum()
    ratio = n_miss / n_data * 100
    asset_name = asset_df[asset_df["Asset_ID"]==asset_id]["Asset_Name"].values[0]
    
    asset_name_lst.append(asset_name)
    missing_ratio_lst.append(ratio)
    
    # print(asset_id,asset_name,n_data,ratio)

mpl.rcParams['figure.figsize'] = (12, 8)
missing_ratio_df = pd.DataFrame(list(zip(asset_name_lst,missing_ratio_lst)), columns = ['Asset_Name','Missing_rate[%]'])
missing_ratio_df = missing_ratio_df.sort_values(['Missing_rate[%]'])
missing_ratio_df.plot.bar(x='Asset_Name',rot = 70)
```

Becareful.  
Bitcoin, Ethereum and Litecoin also has a little missing data.  

```python
mpl.rcParams['figure.figsize'] = (30, 2)
for asset_id in range(14):
    target_df = dropped_sorted_train_df[dropped_sorted_train_df['Asset_ID']==asset_id]
    target_df = target_df.drop(columns=['Asset_ID']).reset_index(drop=True)
    asset_name = asset_df[asset_df["Asset_ID"]==asset_id]["Asset_Name"].values[0]

    ax = plt.axes()
    sns.heatmap(target_df.isna().transpose(),
                cmap="YlGnBu",
                cbar_kws={'label': f'Missing Data'})
    ax.set_title(f'{asset_name}')
    plt.show()
```
