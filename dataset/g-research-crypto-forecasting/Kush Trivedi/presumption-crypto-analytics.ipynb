# What is GluonTS?

The **Gluon Time Series (GluonTS)** toolkit is a **probabilistic time series modeling toolkit** that focuses on **deep learning-based models**.

GluonTS includes utilities for **loading and iterating through time series datasets**, state-of-the-art models that are ready to be trained, and building blocks that allow you to construct your own models and quickly test different solutions. You can use GluonTS to:

1. Use your own data to train and analyze any of the **built-in models**, and **quickly find a solution to your time series problems**.
2. Create bespoke time series models using the given **abstractions** and **building blocks**, then compare them to **baseline techniques quickly**.

**<p id='12'><li>Visualization Helper Function</li></p>**

```python
def plot_prob_forecasts(ts_entry, forecast_entry):
    plot_length = context_length + prediction_length
    prediction_intervals = (50.0, 95.0)
    legend = ["Actual", "Median Prediction"] + [f"{k}% Prediction Interval" for k in prediction_intervals][::-1]

    fig, ax = plt.subplots(1, 1, figsize=(18, 7))
    ts_entry[-plot_length:].plot(ax=ax) 
    forecast_entry.plot(prediction_intervals=prediction_intervals, color='#603F83FF')
    plt.axvline(forecast_entry.start_date, color='#C7D3D4FF', lw=1) 
    plt.grid(which="major")
    plt.legend(legend, loc="upper left")
    plt.show();
```
