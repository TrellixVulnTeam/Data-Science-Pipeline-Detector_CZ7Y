```python
from fbprophet import Prophet
```

```python

train = pd.read_csv("../input/g-research-crypto-forecasting/train.csv")
train['timestamp'] = pd.to_datetime(train['timestamp'],unit='s')
train = train.rename(columns={'timestamp': 'ds',"Target":"y"})
train = train[train["y"].isnull()==0]
train = train[train['ds'] < dt.datetime(2021,6,13)]

def upper_shadow(df):
    return df['High'] - np.maximum(df['Close'], df['Open'])

def lower_shadow(df):
    return np.minimum(df['Close'], df['Open']) - df['Low']

def candle_body(df):
    return np.abs(df.Close-df.Open)

def get_features(df):
    df_feat = df[['Count', 'Open', 'High', 'Low', 'Close', 'Volume', 'VWAP',"y","Asset_ID","ds"]].copy()
    df_feat['Upper_Shadow'] = upper_shadow(df_feat)
    df_feat['Lower_Shadow'] = lower_shadow(df_feat)
    df_feat["body_len"] = candle_body(df_feat)
    return df_feat

train = get_features(train)



m0 = Prophet()
m1 = Prophet()
m2 = Prophet()
m3 = Prophet()
m4 = Prophet()
m5 = Prophet()
m6 = Prophet()
m7 = Prophet()
m8 = Prophet()
m9 = Prophet()
m10 = Prophet()
m11 = Prophet()
m12 = Prophet()
m13 = Prophet()

models = {0: m0,
 1: m1,
 2: m2,
 3: m3,
 4: m4,
 5: m5,
 6: m6,
 7: m7,
 8: m8,
 9: m9,
 10: m10,
 11: m11,
 12: m12,
 13: m13}

for i in range(14):
    models[i].fit(train[train["Asset_ID"]==i])
    pkl_model_path = './prophet-model{0}.pkl'.format(i)
    with open(pkl_model_path, 'wb') as f:
        pickle.dump(models[i], f)
```
