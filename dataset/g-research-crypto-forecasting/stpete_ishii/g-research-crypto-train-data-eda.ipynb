# Period of the train data

```python
import datetime as dt
from datetime import datetime as dt
from datetime import date
from datetime import timedelta

start0 = dt.strptime('2018-01-01 00:01:00','%Y-%m-%d %H:%M:%S')
end0   = dt.strptime('2021-09-21 00:00:00','%Y-%m-%d %H:%M:%S')

print(end0-start0)
seconds0=(end0-start0).days*24*60*60+(end0-start0).seconds
print(seconds0)

time=[]
for i in list(range(0,seconds0+60,60)):
    time+=[start0+timedelta(seconds=i)]
dates0=pd.to_datetime(time,format='%Y-%m-%d %H:%M:%S')
print(dates0)

# train period
dates0_df=pd.DataFrame(dates0)
dates0_df.columns=['ds']
dates0_df
```
