## 1. Load datasets

```python
all_df = pd.read_csv('/kaggle/input/g-research-crypto-forecasting/train.csv')
all_df.head()
```

```python
asset_details = pd.read_csv('/kaggle/input/g-research-crypto-forecasting/asset_details.csv')
asset_details.head()
```

## 2. Merge datasets

```python
all_df = all_df.join(asset_details.set_index('Asset_ID'), on='Asset_ID')
all_df.head()
```

Check if the length of the dataset is the same before and after merging

```python
len(all_df['Asset_ID'].unique()) == len(all_df['Asset_Name'].unique())
```

## 3. Split dataset(Partitioning)

```python
asset_details
```

Split the data according to asset_id, Remove and re-create existing indexes so that indexes are organized in order along with data partitioning. In fact, we will save the data file without index, but for those who will continue to analyze in this kernel, the process is duplicated.

```python
bitcoin_cash_df = all_df.loc[all_df['Asset_ID']==2,:].reset_index(drop=True)
binance_coin_df = all_df.loc[all_df['Asset_ID']==0,:].reset_index(drop=True)
bitcoin_df = all_df.loc[all_df['Asset_ID']==1,:].reset_index(drop=True)
eos_df = all_df.loc[all_df['Asset_ID']==5,:].reset_index(drop=True)
ethereum_classic_df = all_df.loc[all_df['Asset_ID']==7,:].reset_index(drop=True)
ethereum_df = all_df.loc[all_df['Asset_ID']==6,:].reset_index(drop=True)
litecoin_df = all_df.loc[all_df['Asset_ID']==9,:].reset_index(drop=True)
monero_df = all_df.loc[all_df['Asset_ID']==11,:].reset_index(drop=True)
tron_df = all_df.loc[all_df['Asset_ID']==13,:].reset_index(drop=True)
stellar_df = all_df.loc[all_df['Asset_ID']==12,:].reset_index(drop=True)
cardano_df = all_df.loc[all_df['Asset_ID']==3,:].reset_index(drop=True)
iota_df = all_df.loc[all_df['Asset_ID']==8,:].reset_index(drop=True)
maker_df = all_df.loc[all_df['Asset_ID']==10,:].reset_index(drop=True)
dogecoin_df = all_df.loc[all_df['Asset_ID']==4,:].reset_index(drop=True)
```

## 4. Check the partitioned dataset

Check the length of the entire dataset and the partitioned dataset to make sure no assets are missing in the process

```python
len(bitcoin_cash_df) + len(binance_coin_df) + len(bitcoin_df) + len(eos_df) + len(ethereum_classic_df) + len(ethereum_df) + len(litecoin_df) + \
len(monero_df) + len(tron_df) + len(stellar_df) + len(cardano_df) + len(iota_df) + len(maker_df) + len(dogecoin_df)
```
