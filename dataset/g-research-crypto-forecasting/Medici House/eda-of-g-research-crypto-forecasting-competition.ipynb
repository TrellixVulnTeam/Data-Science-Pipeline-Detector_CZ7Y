# Convert timestamp to date format

```python
train_prep1 = train[['timestamp']].applymap(lambda x: time.localtime(x))
train_prep2 = train_prep1[['timestamp']].applymap(lambda x: time.strftime("%Y-%m-%d %H:%M:%S", x))
train['timestamp'] = train_prep2
```

```python
train_2018 = train[train['timestamp']<'2018-02-01 00:00:00']
```

```python
train_2018_ass1 = train_2018[train_2018['Asset_ID']==5]
```

# Market Size of Cryptocurrencies

```python
# Step1 join asset detail information to the train table
train_join = pd.merge(train, asset_details, on = 'Asset_ID', how = 'left')
```

```python
train_join['Market_Size'] = train_join['Close']*train_join['Volume']
```

```python
train_join['year'] = pd.DatetimeIndex(train_join['timestamp']).year
train_join['month'] = pd.DatetimeIndex(train_join['timestamp']).month
train_join['year_month'] = train_join['year'].astype (str) + '/' + train_join['month'].astype (str)
train_join
```

```python
test = train_join
test = pd.DataFrame(test)
test['try'] = test['timestamp']
test['year'] = pd.DatetimeIndex(test['timestamp']).year
test['month'] = pd.DatetimeIndex(test['timestamp']).month#pd.DatetimeIndex(df['Date Attribute']).year
test['year_month'] = test['year'].astype (str) + '/' + test['month'].astype (str)
group_table = test['timestamp'].groupby([test['year_month'],test['Asset_ID']]).max()
```

```python
group_table
```

```python
train_join_join = pd.merge(train_join, group_table, on=['year_month','Asset_ID'], how='left')
train_join_join
```

```python
train_join_join_new = train_join_join[train_join_join['timestamp_x']==train_join_join['timestamp_y']]
train_join_join_new
```

```python
table_sub = train_join_join_new[['year_month','Asset_Name','Market_Size']]
table_sub = table_sub.set_index(['year_month','Asset_Name'])
table_sub
```

```python
table_sub_stack = table_sub.unstack()
table_sub_stack
```

```python
import matplotlib.pyplot as plt
table_sub_stack.plot.bar()
```
