```python
import numpy as np
import pandas as pd

# 14 assets, 60 minutes an hour, 24 hours, 14 days
count = 14*60*24*14

train = pd.read_csv("../input/g-research-crypto-forecasting/supplemental_train.csv", nrows=count)
```

```python
targets = train['Target'].to_numpy()

print("NaN values found:", len(targets[np.isnan(targets)]))

targets = targets[~np.isnan(targets)] # Ignore the few NaN values

max_corr = -np.inf # Track the best correlation over time

# Go through the data with step size 50
for i in range(0, len(targets)-500, 50):
    if i % 1000 == 0:
        print(i, "/", len(targets) - 500, end='\r')
    
    preds = np.zeros(len(targets))
    preds[i:i+500] = 1 # Set 500 consecutive values to 1, starting from i. The rest of the values stay 0.
    
    corr = np.corrcoef(targets, preds)[0, 1] # Calculate the correlation
    
    # Save new best result
    if corr > max_corr:
        max_corr = corr
        print("New best after", i, "minutes: {:.6f}".format(corr))
        
print("Maximum correlation found: {:.6f}".format(max_corr))
```
