But as suggested by [@jpmiller](https://www.kaggle.com/jpmiller) you could also try this:

> You might try flipping the operations here.

> * Do your thing with the data in pandas
> * Save the file back as pd.to_parquet() or feather, hd5

```python
%%time
csvDT = dt.fread("/kaggle/input/g-research-crypto-forecasting/train.csv")
```

```python
%%time 
# Convert datatable to pandas
df = csvDT.to_pandas()
```

# Wrtiting data to diferent file formats

```python
%%time
# Writing dataset as hdf5
data.to_hdf("gCryptoTrain.h5", "gCryptoTrain")
```

```python
%%time
# Writing dataset as feather
data.to_feather("gCryptoTrain.feather")
```

```python
%%time
# Writing dataset as parquet
data.to_parquet("gCryptoTrain.parquet")
```

```python
%%time
# Writing dataset as pickle
data.to_pickle("gCryptoTrain.pkl.gzip")
```

```python
%%time
# Writing dataset as jay
dt.Frame(data).to_jay("gCryptoTrain.jay")
```

# Format: hdf5

```python
%%time

data = pd.read_hdf("gCryptoTrain.h5", "gCryptoTrain")

print("Train size:", data.shape)
```

# Format: feather

```python
%%time

data = pd.read_feather("gCryptoTrain.feather")

print("Train size:", data.shape)
```

# Format: parquet

```python
%%time

data = pd.read_parquet("gCryptoTrain.parquet")

print("Train size:", data.shape)
```

# Format: pickle

```python
%%time

data = pd.read_pickle("gCryptoTrain.pkl.gzip")

print("Train size:", data.shape)
```
