# Visualiztion of a slice of Ethinium Classic Price (CandleStick)

```python
asset_df = pd.read_csv('/kaggle/input/g-research-crypto-forecasting/asset_details.csv')
asset_df
```

```python
etc = train_full[train_full["Asset_ID"]==7].set_index("timestamp") # Asset_ID = 7 for Ethereum Classic
etc_slice = etc[:200]
etc_slice
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Candlestick(x=etc_slice.index, 
                                     open=etc_slice['Open'], 
                                     high=etc_slice['High'], 
                                     low=etc_slice['Low'], 
                                     close=etc_slice['Close'])])
fig.show()
```
