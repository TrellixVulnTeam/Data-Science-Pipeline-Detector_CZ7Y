Taking advantage of usually negative autoccorelation of log returns.
For an explanation (And More !) see my EDA notebook here: https://www.kaggle.com/lucasmorin/crypto-forecasting-intro-eda/

```python
import gresearch_crypto
import numpy as np
import pandas as pd

env = gresearch_crypto.make_env()
iter_test = env.iter_test()

asset_details = pd.read_csv('../input/g-research-crypto-forecasting/asset_details.csv')
#create dictionnary of weights
dict_weights = {}

for i in range(asset_details.shape[0]):
    dict_weights[asset_details.iloc[i,0]] = asset_details.iloc[i,1]
    
for i, (df_test, df_pred) in enumerate(iter_test):
    
    df_test['weights'] = df_test.Asset_ID.map(dict_weights).fillna(0)
    df_test['log_ret'] = np.log(df_test.Close/df_test.Open).fillna(0)
    
    df_pred.Target = -(df_test['log_ret'] - (df_test['weights'] * df_test['log_ret']).sum() / df_test['weights'].sum())

    env.predict(df_pred)
```
