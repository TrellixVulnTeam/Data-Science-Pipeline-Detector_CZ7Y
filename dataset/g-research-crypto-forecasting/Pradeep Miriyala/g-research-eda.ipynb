```python
import numpy as np
import pandas as pd
```

```python
train_data = pd.read_csv('../input/g-research-crypto-forecasting/train.csv')
asset_data = pd.read_csv('../input/g-research-crypto-forecasting/asset_details.csv')
```

## NA Values check

```python
# Percentage of NA vlues in each column
100*train_data.isna().sum()/train_data.shape[0]
```

```python
# We can't work with NA target. These entries shall be removed
train_data.dropna(subset=['Target'],inplace=True)
```

```python
# Recheck NA values.
100*train_data.isna().sum()/train_data.shape[0]
```

## Unique Timestamps

```python
n_unique_ts = len(train_data.timestamp.unique())
print(f'Number of unique timestamps {n_unique_ts}')
print(f'Avearage number of rows per each timestamp {train_data.shape[0]/n_unique_ts}')
```

## Unique Assets

```python
n_unique_as = len(train_data.Asset_ID.unique())
print(f'Number of unique assets {n_unique_as}')
print(f'Avg Number of rows per each asset {train_data.shape[0]/n_unique_as}')
```

## Average number of trades and volume happened at each timestamp

```python
print(f'Avg Number of trades {train_data.Count.mean()}')
print(f'Avg Volume of trades {train_data.Volume.mean()}')
```

## Asset Details

```python
import matplotlib.pyplot as plt
import gc
for i, d in train_data.groupby('Asset_ID'):
    plt.figure(figsize=[16,3])
    plt.subplot(1,4,1)
    plt.plot(d.timestamp, d.Open, d.timestamp, d.Close, d.timestamp, d.VWAP)
    plt.legend(['Open','Close', 'VWAP'])
    plt.xlabel('Time stamp')
    plt.ylabel('USD')
    plt.grid()
    plt.subplot(1,4,2)
    plt.plot(d.timestamp, d.Volume)
    plt.ylabel('Volume')
    plt.xlabel('Time stamp')
    plt.grid()
    plt.subplot(1,4,3)
    plt.plot(d.timestamp, d.High, d.timestamp, d.Low)
    plt.legend(['High', 'Low'])
    plt.ylabel('USD')
    plt.xlabel('Time stamp')
    plt.grid()
    plt.subplot(1,4,4)
    plt.plot(d.timestamp, d.Target)
    plt.ylabel('Target')
    plt.xlabel('Time stamp')
    plt.grid()
    plt.suptitle(asset_data[asset_data.Asset_ID==i]['Asset_Name'].tolist()[0])    
    plt.tight_layout()
    plt.show()
    plt.close('all')
    gc.collect()
```
