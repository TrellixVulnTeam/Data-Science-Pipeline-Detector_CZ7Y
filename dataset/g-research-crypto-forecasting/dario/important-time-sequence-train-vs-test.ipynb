# **Exploring train and test**

Hello,

this is a small exercise to compare test and train dataset.



```python
import pandas as pd
import dask.dataframe as dd
test=dd.read_csv('../input/g-research-crypto-forecasting/example_test.csv')
```

```python
train=dd.read_csv('../input/g-research-crypto-forecasting/train.csv')
```

```python
test['timestamp']=dd.to_datetime(test['timestamp'],unit='s')
test.groupby('Asset_ID').min().compute()
```

```python
train['timestamp']=dd.to_datetime(train['timestamp'],unit='s')
train.groupby('Asset_ID').max().compute()
```
