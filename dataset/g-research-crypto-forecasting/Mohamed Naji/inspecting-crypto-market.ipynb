```python
crypto_df = pd.read_csv(data_folder + 'train.csv')
```

```python
asset_details = pd.read_csv(data_folder + 'asset_details.csv')
```

```python
crypto_df.head()
```

```python
btc = crypto_df.loc[crypto_df.Asset_ID == asset_details.loc[asset_details.Asset_Name=="Bitcoin"].iloc[0, 0]]
btc = btc.set_index("timestamp")
btc_mini = btc.iloc[-200:] #last 200 minutes

import plotly.graph_objects as go

fig = go.Figure(data=[go.Candlestick(x=btc_mini.index, 
                    open=btc_mini['Open'],
                    high=btc_mini['High'], 
                    low=btc_mini['Low'], 
                    close=btc_mini['Close'])])
fig.show()
```

```python
beg_btc = btc.index[0].astype('datetime64[s]')
end_btc = btc.index[-1].astype('datetime64[s]')

print('BTC data goes from ', beg_btc, 'to ', end_btc)
```

*   **Count**: Total number of trades in the time interval (last minute).
*   **Volume**:	Quantity of asset bought or sold, displayed in base currency USD.
*   **VWAP**: The average price of the asset over the time interval, weighted by volume. VWAP is an aggregated form of trade data. 

```python
import matplotlib.pyplot as plt

f = plt.figure(figsize=(15,8))

ax1 = f.add_subplot(311)
plt.plot(btc['VWAP'], label='BTC')
plt.legend()
plt.xlabel('Time')
plt.ylabel('VWAP-Price')

ax2 = f.add_subplot(312)
plt.plot(btc['Volume'], label='BTC', color='r')
plt.legend()
plt.xlabel('Time')
plt.ylabel("Volume(USD)")

ax3 = f.add_subplot(313)
plt.plot(btc['Count'], label='BTC', color='g')
plt.legend()
plt.xlabel('Time')
plt.ylabel("#Trades")

plt.tight_layout()
plt.show()
```
