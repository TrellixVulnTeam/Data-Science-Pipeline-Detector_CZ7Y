```python
import numpy as np
from numpy import loadtxt

d = loadtxt('../input/train.csv', delimiter=',',skiprows=1,converters={1:lambda s:species[str(s)]})
t = loadtxt('../input/test.csv', delimiter=',',skiprows=1)
d = np.delete (d, 0, 1)  # remove row numbers
len(d[0])
```

```python
np.random.shuffle ( d )
# seperate labels and data
labels = d[:,0:1].astype('int')
data = d[:,1:]
means = np.mean(data, axis=0)
stds = np.std(data, axis=0)
data = (data - means)/stds
data = data/10
```

```python
from keras.utils import np_utils

split = int(len(data)*.3)
val_data = data[:split]
val_labels = np_utils.to_categorical(labels[:split])
train_data = data[split:]
train_labels = np_utils.to_categorical(labels[split:])

num_classes = train_labels.shape[1]
```

```python
# try a nn for kicks
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Convolution2D, MaxPooling2D, Flatten
from keras.optimizers import SGD

def model1(ninput, nlabels):
    return Sequential([
        Dense(34, input_dim=ninput, activation='relu'),
        Dropout(.2),
#        Dense(32),
        Dense(nlabels),
        Activation('softmax')
    ])

model = model1(len(train_data[0]), num_classes)
#model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
sgd = SGD(lr=0.1, decay=1e-6, momentum=0.9, nesterov=True)
model.compile(loss='categorical_crossentropy', optimizer=sgd, metrics=['accuracy'])
res = model.fit(train_data, train_labels,
          nb_epoch=400,
          batch_size=200,
          verbose=0)
model.evaluate(val_data, val_labels, batch_size=16, verbose=0)
```

```python
tids = t[:,0:1].astype('int')
tdata = t[:,1:]
tmeans = np.mean(tdata, axis=0)
tstds = np.std(tdata, axis=0)
tdata = (tdata - tmeans)/tstds
tdata = tdata/10

pred = model.predict ( tdata )
```
