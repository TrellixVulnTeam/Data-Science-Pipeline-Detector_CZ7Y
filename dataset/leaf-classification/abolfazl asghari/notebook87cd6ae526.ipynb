```python

import numpy as np
import pandas as pd

# In[4]:


df_test=pd.read_csv('../input/leaf-classification/test.csv.zip')
df_test.head(len(df_test))


# In[25]:


x=df_test.drop(columns=['id'])
y=df_test['id']


# In[26]:


x.shape,y.shape


# In[27]:


x_train=np.array(x)
y_train=np.array(y)


# In[28]:


from sklearn.preprocessing import OneHotEncoder
onehot=OneHotEncoder(sparse=False)
y_train=onehot.fit_transform(y_train.reshape(y_train.shape[0],1))
y_train.shape


# In[29]:


from sklearn.preprocessing import MinMaxScaler
scale=MinMaxScaler()
x_train=scale.fit_transform(x_train)
y_train=scale.fit_transform(y_train)


# In[30]:



import tensorflow as tf


# In[31]:


def model_keras(x_train):
    model_k=tf.keras.models.Sequential()
    #model_k.add(tf.keras.layers.BatchNormalization(input_shape=(594,)))
    model_k.add(tf.keras.layers.Dense(800,activation='relu'))
    model_k.add(tf.keras.layers.Dropout(0.5))
    
    model_k.add(tf.keras.layers.Dense(750,activation='relu'))
    model_k.add(tf.keras.layers.BatchNormalization())
    model_k.add(tf.keras.layers.Dropout(0.2))
    
    model_k.add(tf.keras.layers.Dense(650,activation='relu'))
    model_k.add(tf.keras.layers.BatchNormalization())
    model_k.add(tf.keras.layers.Dropout(0.2))
    
    model_k.add(tf.keras.layers.Dense(600,activation='relu'))
    model_k.add(tf.keras.layers.BatchNormalization())
    model_k.add(tf.keras.layers.Dropout(0.2))
    
        
    #model_k.add(tf.keras.layers.Dense(600,activation='relu'))
    #model_k.add(tf.keras.layers.BatchNormalization())
    #model_k.add(tf.keras.layers.Dropout(0.2))

    model_k.add(tf.keras.layers.Dense(594))
    model_k.add(tf.keras.layers.Activation('sigmoid'))
    model_k.compile(tf.keras.optimizers.Adam(lr=0.005),loss='binary_crossentropy',metrics=['accuracy'])
    return model_k


# In[32]:


model_keras=model_keras(x_train)

model_keras.fit(x_train,y_train,epochs=100,verbose=1)
```
