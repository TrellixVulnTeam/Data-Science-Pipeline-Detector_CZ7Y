```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report
```

```python
test=pd.read_csv('/kaggle/input/leaf-classification/test.csv.zip')
train=pd.read_csv('/kaggle/input/leaf-classification/train.csv.zip')
```

```python
x=train.drop(['id','species'],axis=1)
y=train['species']
```

```python
y
```

```python
encoder=LabelEncoder()
y_fit=encoder.fit(train['species'])
y_label=y_fit.transform(train['species'])
classes=list(y_fit.classes_)
classes
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y_label,test_size=0.2)
```

```python
classifier = RandomForestClassifier(n_estimators = 40)
classifier.fit(x_train,y_train)
```

```python
pred=classifier.predict(x_test)
print(classification_report(y_test,pred))
```

```python
test_data=test.drop('id',axis=1).copy()
```

```python
final_prediction=classifier.predict_proba(test_data)
```

```python
submission=pd.DataFrame(final_prediction,columns=classes)
submission.insert(0,'id',test['id'])
submission.reset_index()
```

```python
submission.to_csv('submission.csv', index = False)
```
