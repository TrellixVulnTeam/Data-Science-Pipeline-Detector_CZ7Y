```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)


import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

```

```python
df944 = pd.read_csv('/kaggle/input/uoflivoutputs/submission_wavenet.csv', dtype={'time': object})
df942 = pd.read_csv('/kaggle/input/uoflivoutputs/gru_preds.csv', dtype={'time': object})
```

```python
df944.head(10)
```

```python
df944.to_csv('df944.csv', index=False)
df942.to_csv('df942.csv', index=False)
```

```python
ens1 = df944.copy()
for ii in ([0.1,0.3,0.5,0.7,0.9]):
    ens1['open_channels'] = ii * df944['open_channels'] + (1.0 - ii) * df942['open_channels']
    ens1['open_channels'] = ens1['open_channels'].apply(lambda x: round(x))
    ens1.to_csv('ens_'+str(ii)+'.csv', index=False)
```
