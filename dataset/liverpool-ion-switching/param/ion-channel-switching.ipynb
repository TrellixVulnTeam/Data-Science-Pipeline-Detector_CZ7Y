```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv("/kaggle/input/liverpool-ion-switching/train.csv") 
# Preview the first 5 lines of the loaded data 
classes = [0,1]
data.head()
```

```python
for clabel in classes:
    # Subset to the airline
    subset = data[data['open_channels'] == clabel]
    print(subset['time'])
    plt.plot(subset['time'], subset['signal'], label = "line 2")
    mng = plt.get_current_fig_manager()
    mng.full_screen_toggle()
    plt.show()
    #plt.show()
    # Draw the density plot
    #sns.distplot(subset['signal'], hist = False, kde = True,kde_kws = {'linewidth': 3},label = clabel)
    
# Plot formatting

plt.legend(prop={'size': 16}, title = 'classes')
plt.title('Density Plot with classes')
plt.xlabel('signal (min)')
plt.ylabel('Density')
```

```python
# Use the 'hue' argument to provide a factor variable
sns.lmplot( x="sepal_length", y="sepal_width", data=data, fit_reg=False, hue='open_channels', legend=False)
 
# Move the legend to an empty part of the plot
plt.legend(loc='lower right')
```

```python
import numpy as np
# import cvxopt
from sklearn.datasets.samples_generator import make_blobs
from sklearn.model_selection import train_test_split
# from matplotlib import pyplot as plt
from sklearn.svm import LinearSVC
from sklearn.metrics import confusion_matrix
import pandas as pd

filePathName = "/kaggle/input/liverpool-ion-switching/train.csv"

df = pd.read_csv(filePathName)

data_array = df.to_numpy()
print(data_array)
x = data_array[:,1].reshape(-1,1)
y = data_array[:,2].reshape(-1,1)

print(x)
X_train, X_test, y_train, y_test = train_test_split(x, y, random_state=0)

svc = LinearSVC()
svc.fit(X_train, y_train)
y_pred = svc.predict(X_test)
confusion_matrix(y_test, y_pred)
```
