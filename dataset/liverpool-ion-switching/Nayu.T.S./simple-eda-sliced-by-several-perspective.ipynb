# signal perspective


I try to see count of open chanels sliced by signal.

I found that open chanels have different trend by batch, so I plot then by batches.

I define signal_class by signal2class.




```python
def signal2class(signal):
    """
    input: signal(float)
    output: signal_class(int)
    --------------
    convert signal to class label.
    signal is...
       ~ -2   -> 1
     -2 ~ 0 -> 2
      0 ~ 2  -> 3
      2 ~ 4  -> 4
      4 ~    -> 5
    """
    if signal < -2:
        return int(1)
    elif signal < 0:
        return int(2)
    elif signal < 2:
        return int(3)
    elif signal < 4:
        return int(4)
    else:
        return int(5)
```

```python
train_df["signal_class"] = train_df["signal"].map(signal2class)
```

```python
plt.figure(figsize=(20, 10))
ax = sns.countplot(x="signal_class", data=train_df[:500000],hue="open_channels")
plt.legend(title="open_channels", bbox_to_anchor=(1, 1), loc='upper left', borderaxespad=0, fontsize=18)
plt.title("Open-Channels Count Plot by signal class of batch1")
```

You can use above　data as csv.

```python
batch1_conditions = (train_df[0:500000]["open_channels"] == 0) | \
                     (train_df[0:500000]["open_channels"] == 1) 
batch2_4_conditions = (train_df[500000:2500000]["open_channels"] == 11)
batch5_conditions = (train_df[2500000:3000000]["open_channels"] == 4) | \
                    (train_df[2500000:3000000]["open_channels"] == 5) | \
                    (train_df[2500000:3000000]["open_channels"] == 6) | \
                    (train_df[2500000:3000000]["open_channels"] == 7) | \
                    (train_df[2500000:3000000]["open_channels"] == 8) 
batch6_9_conditions = (train_df[3000000:4500000]["open_channels"] == 11)
batch10_conditions = (train_df[4500000:5000000]["open_channels"] == 8) | \
                     (train_df[4500000:5000000]["open_channels"] == 9) | \
                     (train_df[4500000:5000000]["open_channels"] == 10) 
```

```python
df_concat = pd.concat([batch1_conditions, batch2_4_conditions, batch5_conditions, batch6_9_conditions, batch10_conditions])
```

```python
train_df_filtered = train_df[df_concat]
train_df_filtered.to_csv("./train_df_filtered.csv", header=False, index=False)
```
