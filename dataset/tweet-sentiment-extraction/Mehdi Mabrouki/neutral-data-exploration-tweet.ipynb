# **Looks like its just half of the neutral data. So why are people assuming that when sentiment is neutral, selected_text = text**

# **Let's take a look at the other neutral data ( where selected_text is different from the original tweet )**

```python
diff = neutral[neutral['text'] != neutral ['selected_text']].reset_index(drop= True)
```

# **O.O**

# **Well, I don't know about you but it looks the same to me... So, let's see why its showing it as different**

# **Let's take the first example**

```python
print(diff.loc[0,'text'],"**", diff.loc[0,'selected_text'])
```

# **Seems the same to me.. What about we print the number of characters**

```python
print(len(diff.loc[0,'text']),"**", len(diff.loc[0,'selected_text']))
```

# **Ahaaa !! There we go, we have an extra character in the original tweet. So, what is it?**

**This is just a simple function that compares two strings characterwise**

```python
def compare(a, b):
    for x, y in zip(a, b):
        if x != y:
            print("Should be ",x," But is ", y)
```

**Alright let's compare'em**

```python
compare(diff.loc[0,'text'], diff.loc[0,'selected_text'])
```

# **Oh.. So obviously there is a " " at the start of the original tweet that does not exist in the selected_text and is messing it all up**

# **So why a lot of people are still using the selected_text = text when sentiment is neutral?**

# **Well, the metric in this competition is the jaccard score, so, let's see if that's changing**

```python
def jaccard(str1, str2): 
    a = set(str1.lower().split()) 
    b = set(str2.lower().split())
    c = a.intersection(b)
    return float(len(c)) / (len(a) + len(b) - len(c))
```

**Alright let's test it out**

```python
jaccard(diff.loc[0,'text'], diff.loc[0,'selected_text'])
```

# **So as you see the jaccard stays completely untouched and is always 1 ( same phrase ), you can check why by searching more about the jaccard score, but nonetheless now we understand why its ( probably ) safe to assume that selected_text is the same as text when the sentiment is neutral**
