## A preliminary EDA before building a model

```python
#import libraries
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import plotly_express as pe
from torchvision.utils import make_grid
import torch
from torchvision.transforms import ToTensor, Resize, Compose
from PIL import Image
import os
import random
# import cv2
```

```python
#read datasets
train_df = pd.read_csv('../input/shopee-product-matching/train.csv')
print(train_df.shape)
train_df.head()
```

```python
# no of similar objects are between 2 and 51
pd.DataFrame(train_df['label_group'].value_counts())
```

```python
#Let's plot histogram of above df
#histogram of labels that have similar products
pe.histogram(train_df['label_group'].value_counts().values)
```

```python
#Let's see some similar images w.r.t to the labels  
def grid_sim_images(df, label='2937985045'):
    label = random.sample(df.label_group.values.tolist(), 1)[0]
    print(label)
    a = train_df[train_df['label_group'] == label]
    print('Titles', a['title'].values)
    print('image phash', a['image_phash'].values)
    imgs = [os.path.join('../input/shopee-product-matching/train_images/',i) for i in a['image'].values.tolist()]
    n =len(imgs)
    tansform = Compose([Resize((500,500)), ToTensor()])
    torch_imgs = []
    for i in imgs:
        torch_imgs.append(tansform(Image.open(i)))
    plt.imshow(make_grid(torch_imgs).permute(1,2,0))

plt.figure(figsize=(60,40))
grid_sim_images(train_df)
```

```python
#Let's see how many similar titles are there in the dataset
pd.DataFrame(train_df['title'].value_counts())
```

```python
#plot histogram for simililar titles 
pe.histogram(train_df['title'].value_counts().values)
```

```python
#get images that have same title (case sensitive)
def grid_sim_images_title(df, title='Emina Glossy Stain'):
    title = random.sample(df.title.values.tolist(), 1)[0]
    print(title)
    a = train_df[train_df['title'] == title]
#     print(a['title'].values)
    imgs = [os.path.join('../input/shopee-product-matching/train_images/',i) for i in a['image'].values.tolist()]
    n =len(imgs)
    if n > 10:
        n = 10
    tansform = Compose([Resize((500,500)), ToTensor()])
    torch_imgs = []
    for i in imgs:
        torch_imgs.append(tansform(Image.open(i)))
    plt.imshow(make_grid(torch_imgs).permute(1,2,0))

grid_sim_images_title(train_df)
```

```python
# see images with similar phash values
def grid_sim_images_phash(df, phash='e992966d4ba49761'):
    image_phash = random.sample(df.image_phash.values.tolist(), 1)[0]
    print(image_phash)
    a = train_df[train_df['image_phash'] == image_phash]
#     print(a['title'].values)
    imgs = [os.path.join('../input/shopee-product-matching/train_images/',i) for i in a['image'].values.tolist()]
    n =len(imgs)
    if n > 10:
        n = 10
    tansform = Compose([Resize((500,500)), ToTensor()])
    torch_imgs = []
    for i in imgs:
        torch_imgs.append(tansform(Image.open(i)))
    plt.imshow(make_grid(torch_imgs).permute(1,2,0))
    
grid_sim_images_phash(train_df)
```
