```python
filename = "/kaggle/input/shopee-product-matching/train.csv"
df = pd.read_csv(filename)
df.head()
```

```python
df_temp=df
```

```python
index=df_temp.index
numberOfRows=len(index)
print(numberOfRows)
```

```python
index=0
data=[]
for index in range(index,numberOfRows+1):    
    while(((df_temp.loc[index,'image_phash'])==(df_temp.loc[index+1,'image_phash'])) is True):        
        data.append([df_temp.loc[index,'posting_id'], df_temp.loc[index,'posting_id'] +" "+ df_temp.loc[index+1,'posting_id']])
        df_results = pd.DataFrame(data, columns=['posting_id', 'matches'])
        index=index+1
```

```python
print(df_results)
```

```python
df_results.to_csv('submission.csv')
```
