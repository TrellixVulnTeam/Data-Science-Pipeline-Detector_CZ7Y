```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
DATA_DIR = '../input/shopee-product-matching/'
```

```python
train_df = pd.read_csv(DATA_DIR + 'train.csv')
train_df.head()
```

```python
train_df.groupby('label_group').count().sort_values('posting_id', ascending=False)
```

```python
top_10_labels = train_df.groupby('label_group').count().sort_values('posting_id', ascending=False).reset_index().iloc[:10, 0].to_list()
```

```python
for row_idx in range(len(train_df[train_df['label_group']==top_10_labels[0]])):
    row = train_df[train_df['label_group']==top_10_labels[0]].iloc[row_idx, :]
    img = row['image']
    title = row['title']
    print(title)
    img_ = cv2.imread(DATA_DIR + 'train_images/' + img)
    print(img)
    plt.imshow(img_)
    plt.show()
```
