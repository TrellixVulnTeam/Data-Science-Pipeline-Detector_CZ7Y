```python
import numpy as np # linear algebra
import pandas as pd
import os
from tqdm import tqdm
import shutil
```

```python
train = pd.read_csv('../input/shopee-product-matching/train.csv')
test = pd.read_csv('../input/shopee-product-matching/test.csv')
sample = pd.read_csv('../input/shopee-product-matching/sample_submission.csv')
```

```python
train=train.drop(['image_phash','title'], axis=1)
train.head()
```

```python
class_img=train['label_group'].drop_duplicates( keep='first', inplace=False)
len(class_img)
```

```python
for fname in tqdm(class_img):
    if not os.path.exists(f"./kaggle/working/class/{fname}"):
         os.makedirs(f"./kaggle/working/class/{fname}")
```

```python
class_file=os.listdir('./kaggle/working/class')
print( type(class_file))

print(len(class_file))             
```

```python
image_path='../input/shopee-product-matching/train_images'
class_path='./kaggle/working/class'
#a=os.listdir(class_path)
#print( len(a))

#移動圖片到指定資料夾
for i in tqdm(range(0,len(train['label_group']))):
    shutil.copy(image_path+'/'+ train['image'][i],class_path+'/'+str(train['label_group'][i]))
```
