```python
df = pd.read_csv('/kaggle/input/shopee-product-matching/train.csv')
```

```python
#df.profile_report()
```

```python
import binascii

df['hash_bin'] = df['image_phash'].apply(lambda x: bin(int(x, base=16))[2:])
df['hash_bin_list'] = df['image_phash'].apply(lambda x: list(map(lambda x: int(x), list(bin(int(x, base=16))[2:]))))
```

```python
test_df = pd.read_csv('/kaggle/input/shopee-product-matching/test.csv')
```

```python
test_df['hash_bin'] = test_df['image_phash'].apply(lambda x: bin(int(x, base=16))[2:])
test_df['hash_bin_list'] = test_df['image_phash'].apply(lambda x: list(map(lambda x: int(x), list(bin(int(x, base=16))[2:]))))
```

```python
import numpy  as np
def dot(x, y):
    return np.dot(np.array(x), np.array(y))
```

```python
for i, y in enumerate(test_df['hash_bin_list']):
    test_df[f'result_{str(i)}'] = test_df['hash_bin_list'].apply(lambda x: dot(x, y))
```

```python
result_dict = dict()
for i, value in enumerate(test_df['posting_id']): 
    result = test_df.sort_values(f'result_{i}')[:10]['posting_id'].values.tolist()
    result.insert(0, value)
    result_dict[value] = [value, ' '.join(result)]
```

```python
result_df = pd.DataFrame.from_dict(result_dict, columns=['posting_id', 'matches'], orient='index')
```

```python
result_df.to_csv(path_or_buf='submission.csv', index=False)
```
