* Image size

```python
image_folder = '../input/shopee-product-matching/train_images/'

sizes = []
limit = 200

for file in os.listdir('../input/shopee-product-matching/train_images/')[:limit]:
    if not file.endswith('.jpg'): raise
    path = os.path.join(image_folder, file)
    img = cv2.imread(path)
    h, w, _ = img.shape
    if h != w: raise
    sizes.append(h)
```

From this small set of 200 images, we can see that the images are square and they have varying dimensions. And the dimensions of different images differ a lot.

```python
np.unique(sizes)
```
