```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
df = pd.read_csv("../input/train-folds/train_folds.csv")
df.head()
```

```python
df['rarity'] = -1
```

```python
df.loc[df.label_group.value_counts()[df.label_group].values<5,'rarity'] = 0
df.loc[(df.label_group.value_counts()[df.label_group].values<10) & (df.label_group.value_counts()[df.label_group].values>=5),'rarity'] = 1
df.loc[(df.label_group.value_counts()[df.label_group].values<15) & (df.label_group.value_counts()[df.label_group].values>=10),'rarity'] = 2
df.loc[(df.label_group.value_counts()[df.label_group].values<20) & (df.label_group.value_counts()[df.label_group].values>=15),'rarity'] = 3
df.loc[(df.label_group.value_counts()[df.label_group].values<25) & (df.label_group.value_counts()[df.label_group].values>=20),'rarity'] = 4
df.loc[(df.label_group.value_counts()[df.label_group].values<35) & (df.label_group.value_counts()[df.label_group].values>=25),'rarity'] = 5
df.loc[(df.label_group.value_counts()[df.label_group].values<40) & (df.label_group.value_counts()[df.label_group].values>=35),'rarity'] = 6
df.loc[(df.label_group.value_counts()[df.label_group].values<45) & (df.label_group.value_counts()[df.label_group].values>=40),'rarity'] = 7
df.loc[(df.label_group.value_counts()[df.label_group].values<=50) & (df.label_group.value_counts()[df.label_group].values>=45),'rarity'] = 8
df.loc[df.label_group.value_counts()[df.label_group].values>50,'rarity'] = 9
```

```python
df.head()
```

```python
df = df.iloc[:,1:]
df.head()
```

```python
df.drop(['fold'], axis=1,inplace = True)
```

```python
from sklearn import model_selection
```

```python
df['fold'] =-1
df = df.sample(frac = 1).reset_index(drop = True)
```

```python
y = df.rarity.values
```

```python
kf = model_selection.StratifiedKFold(n_splits = 5)
```

```python
for f,(t,v) in enumerate(kf.split(X=df,y=y)):
    df.loc[v,'fold'] = f
```

```python
df.head()
```

```python
df.groupby('fold')['rarity'].value_counts()
```

```python
df.to_csv("rarity_folds.csv", index = False)
```
