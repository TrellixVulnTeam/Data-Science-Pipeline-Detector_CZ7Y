# How to convert the escaped bytes back to characters?

Companion notebook for discussion: https://www.kaggle.com/c/shopee-product-matching/discussion/233605

Several examples in the train set have escaped bytes such as "\xc3\x89" in the title column. How do we convert them back to proper characters?

```python
train = pd.read_csv("/kaggle/input/shopee-product-matching/train.csv")
train[train["title"].str.contains("c3")].head()
```

Taking the "title" column as byte-string and decode it in UTF-8 didn't work (returned null values instead).

https://stackoverflow.com/questions/40389764/how-to-translate-bytes-objects-into-literal-strings-in-pandas-dataframe-pytho

```python
train["title"] = train["title"].str.decode("utf-8")
train.info()
```

Reading the CSV in `raw_unicode_escape` encoding didn't work either.

```python
train = pd.read_csv("/kaggle/input/shopee-product-matching/train.csv", encoding="raw_unicode_escape")
train[train["title"].str.contains("c3")].head()
```
