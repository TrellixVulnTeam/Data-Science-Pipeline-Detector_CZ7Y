# Evaluate OCR

```python
# Kaggleのカーネルだとオフラインでpipしなくても使える模様？
import os
import numpy as np
from PIL import Image

# Tesseract
# 参考：https://gammasoft.jp/blog/ocr-by-python/#tesseract
import pyocr
import pyocr.builders
import matplotlib.pyplot as plt

# easyOCR
# 参考:https://gammasoft.jp/blog/ocr-by-python/#tesseract
import easyocr
```

```python
img_path = "../input/shopee-product-matching/train_images/"
files = os.listdir(img_path)

# Tesseract
tools = pyocr.get_available_tools()
tool = tools[0]
builder = pyocr.builders.TextBuilder()
    
# EasyOCR
reader = easyocr.Reader(['en'])
    
for ix, file in enumerate(files):
    # 画像読み込み
    img = Image.open(img_path + file)
    
    # Tesseractの実行
    result_tess = tool.image_to_string(img, lang="eng", builder=builder)
    
    # easyOCR
    result_easy = reader.readtext(img_path + file)
    print(file)
    print("="*50)
    if len(result_tess)==0:
        print("Tesseract : NaN")
    else:
        print("Tesseract↓↓")
        print( result_tess)
        
    print("="*25)
    if len(result_easy)==0:
        print("EasyOCR : NaN")
    else:
        print("EasyOCR↓↓")
        for i in range(len(result_easy)):# EasyOCRは検出した矩形ごとにでてくる
            print(result_easy[i][1])
    
    plt.imshow(np.array(img))
    plt.show()
    if ix==30:
        break
```
