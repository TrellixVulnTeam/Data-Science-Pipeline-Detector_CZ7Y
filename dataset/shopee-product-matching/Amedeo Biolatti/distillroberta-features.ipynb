```python
import os
import gc
import numpy as np
import pandas as pd

import transformers
```

```python
df_train = pd.read_csv("../input/shopee-product-matching/train.csv")
df_train.head()
```

```python
feature_extraction = transformers.pipeline('feature-extraction', 
                                           model="distilroberta-base", 
                                           tokenizer="distilroberta-base", 
                                           device=0)
```

```python
N = 512
text_embedding_avg = []
text_embedding_last = []
for I in range(1000):
    if I * N > df_train.shape[0]:
        break
    print("%d" % I)
    text = df_train[I * N:((I + 1) * N)]['title'].to_list()
    emb = feature_extraction(text)
    text_embedding_avg.append(np.array([np.array(e).mean(0) for e in emb]))
    text_embedding_last.append(np.array([np.array(e)[-1, :] for e in emb]))
    gc.collect()
    pass
text_embedding_avg = np.concatenate(text_embedding_avg, 0)
text_embedding_last = np.concatenate(text_embedding_last, 0)
```

```python
np.save(open('text_embedding_avg.npy', 'wb'), text_embedding_avg)
np.save(open('text_embedding_last.npy', 'wb'), text_embedding_last)
```
