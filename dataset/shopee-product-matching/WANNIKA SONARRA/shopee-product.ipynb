```python
##By Wannika
import os
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import matplotlib.pylab as plt
import cv2
```

```python
train_df = pd.read_csv('../input/shopee-product-matching/train.csv')
train_df.head()
```

```python
label_group = train_df.groupby('label_group',  as_index=False).count()
label_group = label_group.sort_values('image', axis=0, ascending=False)
```

```python
label_group
```

```python
fig = go.Figure([go.Bar(x=label_group.label_group[:20].astype(str).tolist(), y=label_group.image[:20])])
fig.update_layout(title='Top-20 of label group', xaxis_title="Label_group",yaxis_title="Count")
fig.show()
```

```python
def plot_images_group(group):
    
    plot_list = train_df[train_df['label_group'] == group]
    plot_list = plot_list['image'].tolist()
    images_number = len(plot_list)
    size = np.sqrt(images_number)
    if int(size)*int(size) < images_number:
        size = int(size) + 1
        
    plt.figure(figsize=(20, 20))
    
    ind=0
    for image_id in plot_list:
        plt.subplot(size, size, ind + 1)
        image = cv2.imread(os.path.join('../input/shopee-product-matching/train_images/', image_id))
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        plt.imshow(image)
        plt.title(image_id, fontsize=6)
        plt.axis("off")
        ind+=1
    plt.show()
```

```python
test = pd.read_csv('../input/shopee-product-matching/test.csv')
```

```python
check = test.groupby(['title']).count().reset_index()['title'].tolist()
a = []
b = []
for item in check:
    res = test[test['title']== item]['posting_id'].tolist()
    ans = ""
    for id_item in res:
        ans = ans + str(id_item) + " "
    for id_item in res:
        a.append(id_item)
        b.append(ans)
```

```python
submission1 = pd.DataFrame()
submission1['posting_id'] = a
submission1['matches'] = b
submission1
```

```python
check = test.groupby(['image_phash']).count().reset_index()['image_phash'].tolist()
a = []
b = []
for item in check:
    res = test[test['image_phash']== item]['posting_id'].tolist()
    ans = ""
    for id_item in res:
        ans = ans + str(id_item) + " "
    ans = ans[:-1]
    for id_item in res:
        a.append(id_item)
        b.append(ans)
```

```python
submission2 = pd.DataFrame()
submission2['posting_id'] = a
submission2['matches'] = b
submission2
```

```python
submission2 = pd.DataFrame()
submission2['posting_id'] = a
submission2['matches'] = b
submission2
```

```python
sub = pd.merge(submission1, submission2, on='posting_id', how='inner')
sub['list'] = sub['matches_x'] + sub['matches_y']
sub
```

```python
final = []
for index, row in sub.iterrows():
    res = list(set(row['list'].split(' ')))
    ans = ""
    for item in res:
        ans = ans + str(item) + " "
    ans = ans[:-1]
    final.append(ans)
    
submission = pd.DataFrame()
submission['posting_id'] = sub['posting_id']
submission['matches'] = final
submission
```

```python
submission.to_csv('submission.csv', index=False)
```
