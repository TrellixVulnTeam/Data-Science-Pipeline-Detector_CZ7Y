```python
import pandas as pd
import glob
import numpy as np
import os 
import pandas_profiling as pdp
import matplotlib.pyplot as plt
import cv2
from wordcloud import WordCloud, STOPWORDS
import plotly.express as px
# test_df = pd.read_csv("/kaggle/input/shopee-product-matching/train.csv")
test_df = pd.read_csv("/kaggle/input/shopee-product-matching/test.csv")
```

```python
from tqdm import tqdm
submit_df = []
for posting_id,image,image_phash,title in tqdm(zip(test_df.loc[:,"posting_id"],test_df.loc[:,"image"],test_df.loc[:,"image_phash"],test_df.loc[:,"title"])):
    L = list(test_df[test_df["image"] == image]["posting_id"].unique())
    M = list(test_df[test_df["image_phash"] == image_phash]["posting_id"].unique())
    N = list(test_df[test_df["title"] == title]["posting_id"].unique())
    L = list(set(L+M +N))
    L.remove(posting_id)
    matches = ' '.join(map(str, L))
    matches = posting_id +" "+ matches
    submit_df.append([posting_id, matches])
    
submit_df = pd.DataFrame(np.array(submit_df), columns=['posting_id', 'matches'])
    
    
```
