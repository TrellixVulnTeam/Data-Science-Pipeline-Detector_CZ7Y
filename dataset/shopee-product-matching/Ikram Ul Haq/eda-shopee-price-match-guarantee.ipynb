# EDA - SHopee - Price match Guarantee
This notebook provide a good explanation and visualisation of dataset provided in Shopee - Price match Guarantee competition.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

Here we are going too see the meta data of training data. File 'train.csv' contains information about all training examples. Let me explain each column one by one.
- posting_id - the ID code for the posting.
- image - the image id/md5sum.
- image_phash - a perceptual hash of the image.
- title - the product description for the posting.
- label_group - ID code for all postings that map to the same product. Not provided for the test set.

```python
BASE_DIR = '/kaggle/input/shopee-product-matching/'
train_df = pd.read_csv(BASE_DIR + 'train.csv')
train_df
```

**We would like to explore label_group more as it contains information of how many products are in same group or how they match**
Product in row zero fall in group '249114794' let's check what other products fall in same group.

```python
one_group = train_df[train_df['label_group'] == 249114794]
one_group
```

## Total Number of groups for all 34k products

```python
print('Total', len(train_df), 'products fall in', len(train_df['label_group'].unique()), 'groups.')
```

## Same products in one group
**Let's explore two products that are in same group.**

```python
prods = train_df[train_df['label_group'] == 4093212188].reset_index(drop=True)

for i in range(len(prods)):
    print('Product', i)
    image_path = BASE_DIR +'train_images/'+ prods.loc[i, 'image']
    img = mpimg.imread(image_path)
    plt.imshow(img)
    print(prods.loc[i, 'title'])
    plt.show()
```

**In above cell we have seen two products that relies in same group and actually are same products with little bit different information. Our final goal is to recoganise how many products lies in same group**
