### Let us clean the data and create a word cloud

```python
# Convert Text into list 
data  = train.text.tolist()
data[:5]
```

We have lots of special characters, digits and hyper links. 

    Ex: \x89ã¢ (2010) Full\x89ã¢ Streaming - YouTube http://t.co/vVE3UsesGf', 

    Let us remove these by using Regex

```python
# lower the values
data = [str(sentence).lower() for sentence in data]
data[:1]
```

```python
import re

# remove the https/www or http
data = [re.sub(r"https|www|http\S+", "", sent) for sent in data]
data[:50]
```

```python
# remove speacial characters
data = [re.sub('\S*#|$|@|=|>|!|:\S*\s?', '', sent) for sent in data]
data[:50]
```

```python
# Substituting multiple spaces with single space
data = [re.sub(r'\s+', ' ', sent) for sent in data]
data[:1]
```

```python
# remove the digit
data = [re.sub(r"\d", "", sent) for sent in data]
data = [re.sub(r'\^[a-zA-Z0-9]\s]', "", sent) for sent in data]
data[:5]
```

```python
train['clean_data'] = pd.DataFrame(data)
```

```python
train.info()
```

```python
# importing all necessery modules 
from wordcloud import WordCloud, STOPWORDS 
```

```python
# code refereed from internet 
comment_words = ' '
stopwords = set(STOPWORDS) 

for data in train.clean_data: 

    # split the value 
    tokens = data.split() 
    for i in range(len(tokens)): 
        tokens[i] = tokens[i].lower() 
          
    for words in tokens: 
        comment_words = comment_words + words + ' '
  

wordcloud = WordCloud(width = 500, height = 500, 
                background_color ='white', 
                stopwords = stopwords, 
                min_font_size = 8).generate(comment_words) 
                  
plt.figure(figsize = (10, 10)) 
plt.imshow(wordcloud) 
plt.show() 
```
