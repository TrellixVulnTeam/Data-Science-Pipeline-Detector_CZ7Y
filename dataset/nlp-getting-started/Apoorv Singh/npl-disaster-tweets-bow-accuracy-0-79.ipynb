```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
train=pd.read_csv("../input/nlp-getting-started/train.csv")
test=pd.read_csv("../input/nlp-getting-started/test.csv")
```

```python
train.head()
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
def prepare_data(train_docs, test_docs, mode):
  # create the tokenizer
  tokenizer = Tokenizer()
  # fit the tokenizer on the documents
  tokenizer.fit_on_texts(train_docs)
  # encode training data set
  Xtrain = tokenizer.texts_to_matrix(train_docs, mode=mode)
  # encode training data set
  Xtest = tokenizer.texts_to_matrix(test_docs, mode=mode)
  return Xtrain, Xtest
```

```python
from sklearn.model_selection import train_test_split
y=train['target']
t1,t2=prepare_data(train['text'],test['text'],'count')
trainx,testx,trainy,testy = train_test_split(t1,y,test_size=0.2,random_state=2)
```

```python
from sklearn.metrics import accuracy_score

def model(model):
    model.fit(trainx,trainy)
    modelpred=model.predict(testx)
    return(accuracy_score(modelpred, testy)),modelpred
```

```python
from sklearn.neural_network import MLPClassifier

model = MLPClassifier(hidden_layer_sizes=(10,4,),verbose=2,early_stopping=True)
model.fit(trainx,trainy)
modelpred=model.predict(testx)
accuracy_score(modelpred, testy)
```

```python
test_predictions = model.predict(t2)
```

```python
submission = pd.DataFrame({
        "id": test['id'],
        "target":test_predictions
    })
submission.to_csv('submission.csv', index=False)
```
