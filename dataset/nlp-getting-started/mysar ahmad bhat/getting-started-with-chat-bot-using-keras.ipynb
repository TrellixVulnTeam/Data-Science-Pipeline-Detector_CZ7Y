<a id="1.1"></a>
<p style = "font-size:30px; color:#331340 ; text-align: left; border-radius: 5px 5px; padding: 5px" ><strong>Concept</strong></p>

<div style="background: #DCDCDC">
<p style = "font-size:18px; ; text-align: center" >Before jumping into the coding section, first, we need to understand some design concepts. Since we are going to develop a deep learning based model, we need data to train our model. But we are not going to gather or download any large dataset since this is a simple chatbot. We can just create our own dataset in order to train the model. To create this dataset, we need to understand what are the intents that we are going to train. An ‚Äúintent‚Äù is the intention of the user interacting with a chatbot or the intention behind each message that the chatbot receives from a particular user. According to the domain that you are developing a chatbot solution, these intents may vary from one chatbot solution to another. Therefore it is important to understand the right intents for your chatbot with relevance to the domain that you are going to work with.
<br>
Then why it needs to define these intents? That‚Äôs a very important point to understand. In order to answer questions, search from domain knowledge base and perform various other tasks to continue conversations with the user, your chatbot really needs to understand what the users say or what they intend to do. That‚Äôs why your chatbot needs to understand intents behind the user messages (to identify user‚Äôs intention).
<br>
How can you make your chatbot understand intents in order to make users feel like it knows what they want and provide accurate responses. The strategy here is to define different intents and make training samples for those intents and train your chatbot model with those training sample data as model training data (X) and intents as model training categories (Y)..</p>
</div>

<a id="1.2"></a>
<p style = "font-size:30px; color:#331340 ; text-align: left; border-radius: 5px 5px; padding: 5px" ><strong>Import libraries</strong></p>

```python
import json
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense,Embedding,GlobalAveragePooling1D
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from sklearn.preprocessing import LabelEncoder
```

<a id="1.3"></a>
<p style = "font-size:30px; color:#331340 ; text-align: left; border-radius: 5px 5px; padding: 5px" ><strong>Define Intents</strong></p>

<div style="background: #DCDCDC">
<p style = "font-size:18px; ; text-align: center" >I will define few simple intents and bunch of messages that corresponds to those intents and also map some responses according to each intent category. I will create a JSON file named ‚Äúintents.json‚Äù including these data as follows.</p>
</div>

```python
with open('../input/intents/intents.json') as file:
    data = json.load(file)
```

```python
data
```

<a id="1.4"></a>
<p style = "font-size:30px; color:#331340 ; text-align: left; border-radius: 5px 5px; padding: 5px" ><strong>Data Prepration</strong></p>

<div style="background: #DCDCDC">
<p style = "font-size:18px; ; text-align: center" >Now we load the json file and extract the required data..</p>
</div>

```python
training_sentences = []
training_labels = []
labels = []
responses = []
for intent in data['intents']:
    for patterns in intent['patterns']:
        training_sentences.append(patterns)
        training_labels.append(intent['tag'])
    responses.append(intent['responses'])
    
    if intent['tag'] not in labels:
        labels.append(intent['tag'])
        
        
    
```

<div style="background: #DCDCDC">
<p style = "font-size:18px; ; text-align: center" >The variable ‚Äútraining_sentences‚Äù holds all the training data (which are the sample messages in each intent category) and the ‚Äútraining_labels‚Äù variable holds all the target labels correspond to each training data.
<br>
Then we use ‚ÄúLabelEncoder()‚Äù function provided by scikit-learn to convert the target labels into a model understandable form.

.</p>
</div>

```python
encoder= LabelEncoder()
encoder.fit(training_labels)
training_labels = encoder.transform(training_labels)
```

<div style="background: #DCDCDC">
<p style = "font-size:18px; ; text-align: center" >Next, we vectorize our text data corpus by using the ‚ÄúTokenizer‚Äù class and it allows us to limit our vocabulary size up to some defined number. When we use this class for the text pre-processing task, by default all punctuations will be removed, turning the texts into space-separated sequences of words, and these sequences are then split into lists of tokens. They will then be indexed or vectorized. We can also add ‚Äúoov_token‚Äù which is a value for ‚Äúout of token‚Äù to deal with out of vocabulary words(tokens) at inference time.</p>
</div>

```python
vocab_size = 1000
embedding_dim = 16
max_len = 20
oov_token = "<OOV>"

tokenizer = Tokenizer(num_words=vocab_size, oov_token=oov_token)
tokenizer.fit_on_texts(training_sentences)
word_index = tokenizer.word_index
sequences = tokenizer.texts_to_sequences(training_sentences)
padded_sequences = pad_sequences(sequences, truncating='post', maxlen=max_len)
```


<div style="background: #DCDCDC">
<p style = "font-size:18px; ; text-align: center" >The ‚Äúpad_sequences‚Äù method is used to make all the training text sequences into the same size..</p>
</div>

<a id="1.5"></a>
<p style = "font-size:30px; color:#331340 ; text-align: left; border-radius: 5px 5px; padding: 5px" ><strong>Model Training</strong></p>


<div style="background: #DCDCDC">
<p style = "font-size:18px; ; text-align: center" >Let‚Äôs define our Neural Network architecture for the proposed model and for that we use the ‚ÄúSequential‚Äù model class of Keras.</p>
</div>


```python
num_classes = len(labels)
model = Sequential()
model.add(Embedding(vocab_size, embedding_dim, input_length=max_len))
model.add(GlobalAveragePooling1D())
model.add(Dense(16, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(num_classes, activation='softmax'))

model.compile(loss='sparse_categorical_crossentropy', 
              optimizer='adam', metrics=['accuracy'])

model.summary()
```


<div style="background: #DCDCDC">
<p style = "font-size:18px; ; text-align: center" >Now we are ready to train our model. Simply we can call the ‚Äúfit‚Äù method with training data and labels.s.</p>
</div>


```python
epochs = 500
history = model.fit(padded_sequences, np.array(training_labels), epochs=epochs)
```

<a id="1.6"></a>
<p style = "font-size:30px; color:#331340 ; text-align: left; border-radius: 5px 5px; padding: 5px" ><strong>Inference</strong></p>

<div style="background: #DCDCDC">
<p style = "font-size:18px; ; text-align: center" >Okay!!!! now it‚Äôs time to check how our model performs. üòä
<br>
We are going to implement a chat function to engage with a real user. When a new user message is received, the chatbot will calculate the similarity between the new text sequence and training data. Considering the confidence scores got for each category, it categorizes the user message to an intent with the highest confidence score.</p>
</div>


```python
while True:
    inp = input()
    if inp.lower() == "exit":
        break
    result = model.predict(keras.preprocessing.sequence.pad_sequences(tokenizer.texts_to_sequences([inp]),
                                             truncating='post', maxlen=max_len))
    tag = encoder.inverse_transform([np.argmax(result)])

    for i in data['intents']:
        if i['tag'] == tag:
            print("ChatBot:"  , np.random.choice(i['responses']))

        
```
