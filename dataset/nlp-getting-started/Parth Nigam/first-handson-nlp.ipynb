```python
dist_tweets= train[train['target']==1]['text']
dist_tweets.head()
```

```python
nondist_tweets= train[train['target']!=1]['text']
nondist_tweets.head()
```

```python
stopwords = set(STOPWORDS)

def show_wordcloud(data, title = None):
    wordcloud = WordCloud(
        background_color='white',
        stopwords=stopwords,
        max_words=200,
        max_font_size=40, 
        scale=3,
        random_state=1 # chosen at random by flipping a coin; it was heads
    ).generate(str(data))

    fig = plt.figure(1, figsize=(12, 12))
    plt.axis('off')
    if title: 
        fig.suptitle(title, fontsize=20)
        fig.subplots_adjust(top=2.3)

    plt.imshow(wordcloud)
    plt.show()
show_wordcloud(dist_tweets,title='Disaster Tweets')
show_wordcloud(nondist_tweets,title='Non Disaster Tweets')
```

```python
text=list(train['text'])
```

```python
wordnet= WordNetLemmatizer()
stopwords.words('english')
```

```python
corpus=[]
for i in range(len(text)):
    review= re.sub('[^a-zA-Z]','',text[i])
    review= review.lower()
    review= [wordnet.lemmatize(word) for word in review if not word in set(stopwords.words('english'))]
    review= ' '.join(review)    
    corpus.append(review)
```
