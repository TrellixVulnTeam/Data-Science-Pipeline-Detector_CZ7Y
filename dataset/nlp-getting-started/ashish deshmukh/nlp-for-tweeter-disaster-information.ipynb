```python
#Cleaning the Tweets by Removing Panctuation marks,Symbols,URLS,Stopwords,
import nltk
from nltk.tokenize import sent_tokenize,word_tokenize
nltk.download('punkt')
```

```python
x=train.text
```

```python
word_tokenize(x[1])
```

```python
import re
```

```python
x=[re.sub('http[s]?://\S+#.','',sent)for sent in x]
```

```python
from nltk.corpus import stopwords
#nltk.download('stopwords')
```

```python
stop_word=stopwords.words('english')
```

```python
import string
```

```python
from string import punctuation
```

```python
print(list(punctuation))
```

```python
final_stop=stop_word+list(punctuation)+["...","''","===","^^","#","http"]
```

```python
#Normalising the case
x=[word.lower() for word in x]
```

```python
x=[word_tokenize(sent) for sent in x]
```

```python
def stop(sent):
    return [word for word in sent if word not in final_stop]
```

```python
stop(x[0])
```

```python
x=[stop(sent) for sent in x]
```

```python
x=[" ".join(word) for word in x]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
```

```python
tfidfv=TfidfVectorizer()
```

```python
X=x
y=train.target
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train=tfidfv.fit_transform(X_train)
```

```python
X_test=tfidfv.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
Log_reg=LogisticRegression()
```

```python
Log_reg.fit(X_train,y_train)
```

```python
pred=Log_reg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,f1_score,confusion_matrix
```

```python
print(f"{'Model'}\n")
print(f"Classication Report\n \n{classification_report(pred,y_test)}")
print(f"Accuracy Score using Model {accuracy_score(pred,y_test)}\n")
print(f"Confusion Matrix \n {confusion_matrix(pred,y_test)}")
```

From Above we can see that our Accuracy is Very good but we can improve it by using Advanced Algorithms

```python
def Results(pred,y_test):
    #print(f"{a}\n")
    print(f"Classication Report\n \n{classification_report(pred,y_test)}")
    print(f"Accuracy Score using Model {accuracy_score(pred,y_test)}\n")
    print(f"Confusion Matrix \n {confusion_matrix(pred,y_test)}")
```
