```python
df = pd.read_csv('/kaggle/input/nlp-getting-started/train.csv')
df.head()
```

```python
df1 = pd.DataFrame(df['text'])
df1
```

## Date

```python
import re
def date(text):
    pattern = re.compile(r'\d{2}[-/.]?\d{2}[-/.]?\d{4}')
    text = pattern.findall(text)
    return text
df1['Date'] = df1['text'].apply(lambda x : date(x))
df1
```

## #hashtag

```python
def hashtag(text):
    pattern = re.compile(r'#\w*')
    text = pattern.findall(text)
    return text
df1['Hashtag'] = df1['text'].apply(lambda x : hashtag(x))
df1
```

## @Mention

```python
def mention(text):
    pattern = re.compile(r'@\w*')
    text = pattern.findall(text)
    return text
df1['Mention'] = df1['text'].apply(lambda x : mention(x))
df1
```

## Time

```python
def time(text):
    pattern = re.compile(r'\d{2}:\d{2} \w*')
    text = pattern.findall(text)
    return text
df1['time'] = df1['text'].apply(lambda x : time(x))
df1
```

## URLs

```python
def link(text):
    pattern = re.compile(r'http[s]?://[^\s]+')
    text = pattern.findall(text)
    return text
df1['URLs'] = df1['text'].apply(lambda x : link(x))
df1
```

## Years

```python
def year(text):
    pattern = re.compile(r'\d{4}')
    text = pattern.findall(text)
    return text
df1['Year'] = df1['text'].apply(lambda x : year(x))
df1
```

## Alphanumeric

```python
def alphanumeric(text):
    pattern = re.compile(r'^\w*')
    text = pattern.findall(text)
    return text
df1['Alphanumeric'] = df1['text'].apply(lambda x: alphanumeric(x))
df1
```

## %Percentage

```python
def percentage(text):
    pattern = re.compile(r'[\d.\d]+%')
    text = pattern.findall(text)
    return text
df1['%'] = df1['text'].apply(lambda x: percentage(x))
df1    
```

# Backreferencing

```python
## convert the date dd/mm/yyyy into yyyy-mm-dd
string = """
Yesterday is 08/03/2022
Today is 09/03/2022
Tommarow is 10/03/2022"""

pattern = re.compile(r'(\d{2})\/(\d{2})\/(\d{4})') ## make group and remove / using \
print("Groups :", pattern.findall(string))
print(pattern.sub(r'\3-\2-\1', string)) ## convert by using group indexing
```

# Named Group 


```python
string1 = "Shrikant Shejwal"
pattern = re.compile(r'(?P<first>\w+) (?P<last>\w+)') ## make group by using ?P<group name>
print(pattern.findall(string1))

##check the groups
match = pattern.match(string1)
print(match.group('first'))
print(match.group('last'))

## swap the groups with group indexing as well as group names
print(pattern.sub(r'\2 \1', string1))
## or
print(pattern.sub(r'\g<last> \g<first>', string1))
```

# Boundry

```python
string3 = 'The energy of god is always working through me.this is a way to keep your mind positive'
pattern = re.compile(r'way')
print("without boundry =>", pattern.findall(string3))

## apply boundry - it can consider only word present individual
pattern1 = re.compile(r'\bway')
print("with boundry =>", pattern1.findall(string3))
```

# Look ahead

```python
## positive look ahead
string4 = 'the energy of god is always working through me. the this is a way to keep your mind positive'
pattern = re.compile(r'the(?=\senergy)') ## find "The" word followed by next word energy
print('Positive look ahead "the" word followed by next word energy =>',pattern.findall(string4))

## negative look ahead
pattern1 = re.compile(r'the(?!\senergy)') ## find "The" word followed by next word energy
print('Negative look ahead "the" word is not followed by next word energy =>',pattern1.findall(string4))
```
