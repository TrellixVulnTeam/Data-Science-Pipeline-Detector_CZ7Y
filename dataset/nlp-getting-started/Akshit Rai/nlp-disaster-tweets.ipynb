    <h1><i><font face='Courier' size = 10>Importing Libraries</font></i></h1>

```python
from nltk.corpus import stopwords
from sklearn.linear_model import LogisticRegression
```

```python
import numpy as np
import pandas as pd
```

```python
import string
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.naive_bayes import MultinomialNB
```

<html>
    <h1><i><font face='Courier' size = 10>Data Cleaning and the Model</font></i></h1>

```python
train = pd.read_csv('../input/nlp-getting-started/train.csv')
test = pd.read_csv('../input/nlp-getting-started/test.csv')
```

<html>
    <font face='Courier' size = 3>Now Lets Create the Function for removing stopwords and punctuations</font>

```python
def cleaner(x):
    return [a for a in (''.join([a for a in x if a not in string.punctuation])).split() if a.lower() not in stopwords.words('english')]
```

<html>
        <font face='Courier' size = 3>We are ready to make Pipeline for this Model</font>

```python
lr = LogisticRegression(solver='liblinear', random_state=777) 
```

```python
Pipe = Pipeline([
    ('bow',CountVectorizer(analyzer=cleaner)),
    ('tfidf',TfidfTransformer()),
    ('classifier',MultinomialNB())
])
```

```python
train.drop('id',inplace = True,axis = 1)
test1 = test.drop('id',axis = 1)
```

```python
train.fillna('None',inplace=True)
test1.fillna('None',inplace=True)
test.fillna('None',inplace=True)
```

```python
train
```

```python
Pipe.fit(train[['text','keyword','location']].values,train['target'])
```

<html>
    <h1><i><font face='Courier' size = 10>Prediction</font></i></h1>

```python
pred2 = Pipe.predict(test[['text','keyword','location']].values)

```

```python
submit_pred = pd.Series(pred2)
```
