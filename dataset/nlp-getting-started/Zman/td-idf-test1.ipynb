```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
```

```python
test = pd.read_csv("../input/nlp-getting-started/test.csv")
train = pd.read_csv("../input/nlp-getting-started/train.csv")
```

```python
print(train.shape)
print(train['target'].unique())
train.head()
```

```python
train_text=train['text']
test_text=test['text']

TFvec=TfidfVectorizer(max_features=10000).fit(train_text)

X_train=TFvec.transform(train_text)
X_test=TFvec.transform(test_text)
Y_train= train["target"]

TFvec.vocabulary_
#字典
```

```python
X_train.shape
```

```python
print(X_train[50].toarray())
```

```python
clf= LogisticRegression(random_state=0)
clf.fit(X_train,Y_train)
y_pred=clf.predict(X_test)
```

```python
y_pred[:50]
```

```python
ss=pd.read_csv("../input/nlp-getting-started/sample_submission.csv")
ss["target"]=y_pred
ss.to_csv("test1.csv",index=False)
```
