```python
train_df = pd.read_csv('/kaggle/input/nlp-getting-started/train.csv')
```

```python
train_df.count()
```

# **Dropping duplicates and NaN from the dataframe**

```python
train_df = train_df.dropna()
train_df = train_df.drop_duplicates()
```

```python
train_df.count()
```

```python
train_df.head()
```

# **Apply tweet preprocessing first. Define a preprocess function and use pandas apply to apply it on each value of 'text'**

```python
import preprocessor as p

def preprocess_tweet(row):
    text = row['text']
    text = p.clean(text)
    return text
```

```python
train_df['text'] = train_df.apply(preprocess_tweet, axis=1)
```

# **Tweet has been cleaned to normal text.**

```python
train_df.head()
```

# **Now we can apply normal text preprocessing like**
* Lowercasing
* Punctuation Removal
* Replace extra white spaces
* Stopwords removal**


```python
from gensim.parsing.preprocessing import remove_stopwords

def stopword_removal(row):
    text = row['text']
    text = remove_stopwords(text)
    return text
```

```python
train_df['text'] = train_df.apply(stopword_removal, axis=1)
```

```python
train_df.head()
```

# **Remove extra white spaces, punctuation and apply lower casing**

```python
train_df['text'] = train_df['text'].str.lower().str.replace('[^\w\s]',' ').str.replace('\s\s+', ' ')
```

```python
train_df.head()
```

1. # **Now input tweet has been pre-processed and lets find features.**

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tweets = train_df['text']
vectorizer = TfidfVectorizer(stop_words='english')

# Learn vocabulary from sentences. 
vectorizer.fit(tweets)

# Get vocabularies.
vectorizer.vocabulary_
```

```python
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer

stopwords = stopwords.words('english')

print(stopwords)

count_vector = CountVectorizer(token_pattern=r'\w{1,}', ngram_range=(1, 2), stop_words = stopwords)
```

```python
from sklearn.model_selection import train_test_split
X = train_df.text
y = train_df.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

clf = LogisticRegression()
pipe = Pipeline([
    ('count_vector', CountVectorizer()),  
    ('clf', LogisticRegression()) 
])
pipe.fit(X_train,y_train)
```

```python
from sklearn import metrics
predicted = pipe.predict(X_test)
```

```python
print("accuracy :",  metrics.accuracy_score(predicted, y_test))
```

```python
submission = pd.read_csv('/kaggle/input/nlp-getting-started/sample_submission.csv')
submission["target"] = pipe.predict(test_df.text)
submission.to_csv("submission.csv", index=False)
```
