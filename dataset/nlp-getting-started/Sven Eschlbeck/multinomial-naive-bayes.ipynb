# **NLP with Multinomial Naive Bayes**

#### Loading and Displaying data

```python
import pandas as pd
```

```python
df = pd.read_csv("../input/nlp-getting-started/train.csv", encoding = "utf-8")
df.head()
```

```python
df.dtypes
```

#### Simplifying and Cleaning data

```python
df["keyword"] = df["keyword"].astype("string")
df["location"] = df["location"].astype("string")
df["text"] = df["text"].astype("string")
```

```python
df["keyword"] = df["keyword"].fillna(" ")
df["location"] = df["location"].fillna(" ")
```

```python
df["message"] = df["keyword"] + " " + df["location"] + " " + df["text"]
df.head()
```

#### Preparing data

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
```

```python
X = df["message"]
y = df["target"]

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
```

```python
X_train
```

```python
cv = CountVectorizer()

cv.fit(X_train)

X_train = cv.transform(X_train)
X_test = cv.transform(X_test)

print(X_train)
```

```python
X_train.shape
```

### **Multinomial Naive Bayes**

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(X_train, y_train)

print(model.score(X_test, y_test))
```
