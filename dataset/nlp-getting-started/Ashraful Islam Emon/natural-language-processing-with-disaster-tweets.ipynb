# Import DATA from kaggle 

```python
df1= pd.read_csv('../input/nlp-getting-started/train.csv')
df2= pd.read_csv('../input/nlp-getting-started/test.csv')
```

```python
df1
```

# REMOVE duplicates values 

```python
df1.drop_duplicates(inplace=True)
```

```python
df1
```

# Import DATA

```python
x=df1['text']
y=df1['target']

x1=df2['text']
```

# preprocessing
**Feature_extraction**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf= TfidfVectorizer()
x_train=tf.fit_transform(x)
```

# Classifier

```python
from sklearn.svm import LinearSVC 
model = LinearSVC()
model.fit(x_train,y)
```

# Predicting 

```python
cv_email = tf.transform(x1)
pred = model.predict(cv_email)
```

```python
pred
```

# Submission data

```python
submission =pd.DataFrame({
     'id': df2['id'],
     'target': pred
     })
submission.to_csv("new1_nlp1.csv",index=False)
```
