### Importing libraries
We will import the libraries we use in this kernel :
* Essential python DS libraries like numpy, pandas
* Autogluon module: TextPredictor

```python
from autogluon.text import TextPredictor #TextPredictor modul for text prediction
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

#### Loading  the Data 

```python
train_df = pd.read_csv("/kaggle/input/nlp-getting-started/train.csv") #train data 
test_df = pd.read_csv("/kaggle/input/nlp-getting-started/test.csv") # test data 
sample_submission = pd.read_csv("../input/nlp-getting-started/sample_submission.csv") # sample submission 
```

#### Training the model

```python
predictor = TextPredictor(label='target', eval_metric='acc', path='./ag_sst') #setting the label and evaluation metric 
predictor.fit(train_df, time_limit=200) # seting time to be 200 sec  ( this is only for demonstration )
```

#### Generating predictions for the test data  

```python
predictions =  predictor.predict(test_df)
print("Predictions:  \n", predictions)
```

#### Intermediate Training Results¬∂


```python
predictor.results.tail(5)
```

#### Making a submission 

```python
submission = test_df[['id']]
submission = pd.concat([submission, predictions], axis=1)
submission.to_csv('submission.csv', index=False)
submission.head()
```
