```python
import numpy as np 
import pandas as pd 
import os
from collections import defaultdict
import seaborn as sns
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
stop = set(stopwords.words('english'))
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score,f1_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

**Loading Data**

```python
train_df = pd.read_csv("/kaggle/input/nlp-getting-started/train.csv")
test_df = pd.read_csv("/kaggle/input/nlp-getting-started/test.csv")
sub_df = pd.read_csv("/kaggle/input/nlp-getting-started/sample_submission.csv")
```

```python
def create_corpus(target):
    corpus=[]
    
    for x in train_df[train_df['target']==target]['text'].str.split():
        for i in x:
            corpus.append(i)
    return corpus
```

```python
corpus=create_corpus(0)

dic=defaultdict(int)
for word in corpus:
    if word in stop:
        dic[word]+=1
        
top=sorted(dic.items(), key=lambda x:x[1],reverse=True)[:10] 

```

```python
for k in range(0,len(top)):
    print(top[k][0])
```

```python
x,y = zip(*top)
print(x)
plt.bar(x,y)
```

```python
x = train_df["text"]
y = train_df["target"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
X_train
```

```python
vect = CountVectorizer(stop_words = 'english')

x_train_cv = vect.fit_transform(X_train)
x_test_cv = vect.transform(X_test)
```

```python
x_train_cv
```

```python
clf = MultinomialNB()
clf.fit(x_train_cv, y_train)
```

```python
pred = clf.predict(x_test_cv)
```

```python
pred
```

```python
y_test
```

```python
# y_pred = model.predict(X_test)

f1score = f1_score(y_test,pred)
print(f"Model Score: {f1score * 100} %")
```

```python
confusion_matrix(y_test, pred)
```

```python
accuracy_score(y_test,pred)
```

```python
y_test = test_df["text"]
y_test_cv = vect.transform(y_test)
preds = clf.predict(y_test_cv)
```

```python
sub_df["target"] = preds
sub_df.to_csv("submission.csv",index=False)
```
