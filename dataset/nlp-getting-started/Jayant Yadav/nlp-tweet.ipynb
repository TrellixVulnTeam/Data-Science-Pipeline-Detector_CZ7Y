```python
train_df=pd.read_csv('../input/nlp-getting-started/train.csv')
test_df=pd.read_csv('../input/nlp-getting-started/test.csv')
```

```python
train_df.head()
```

```python
print(f'number of rows:{train_df.shape[0]}')
print('-'*80)
print(f'data types :\n{train_df.dtypes}')
print('-'*80)
c=train_df.isnull().sum()
print(f'columns with null values:\n{c[c>0]}')
```

```python
import spacy

nlp=spacy.load('en_core_web_lg')
```

```python
with nlp.disable_pipes():
    doc_vectors=np.array([nlp(text).vector for text in train_df.text ])
```

```python
doc_vectors.shape
```

```python
X=doc_vectors
y=train_df.target
```

```python
from sklearn.svm import LinearSVC
svc=LinearSVC(random_state=1,dual=False,max_iter=1000)
svc.fit(X,y)
```

```python
yhat=svc.predict(X)
yhat
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(yhat,y))
```

```python
with nlp.disable_pipes():
    test_vec=np.array([nlp(text).vector for text in test_df.text])
```

```python
target=svc.predict(test_vec)
```

```python
target
```

```python
output=pd.DataFrame({'id':test_df.id,'target':target})
output.to_csv('submit.csv',index=False)
```
