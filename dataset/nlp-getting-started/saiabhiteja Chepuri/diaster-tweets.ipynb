```python
import texthero as hero
from texthero import preprocessing as ppe
```

```python
custom_pipeline = [ppe.fillna,
                   ppe.lowercase,
                   ppe.remove_punctuation,
                   ppe.remove_stopwords,
                   ppe.remove_digits,
                   ppe.remove_diacritics,
                   ppe.remove_round_brackets,
                   ppe.remove_html_tags,
                   ppe.remove_urls,
                   ppe.remove_whitespace,
                   ppe.remove_brackets,
                   ppe.remove_angle_brackets,
                   ppe.remove_curly_brackets,
                   ppe.remove_square_brackets,
                   ppe.stem
                  ]
```

```python
df['clean_text'] = hero.clean(df['text'], custom_pipeline)
dftest['clean_text'] = hero.clean(dftest['text'], custom_pipeline)
```

```python
vector  = hero.CountVectorizer()
x = vector.fit_transform(df['clean_text'])
xtest = vector.transform(dftest['clean_text'])
```

```python
y =df['target'].values
```

<h2> Model </h2>

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
lr = LogisticRegression()
cv = cross_val_score(lr,x,y,cv = 10)
cv
```

```python
from sklearn.model_selection import GridSearchCV
gd = GridSearchCV(
    estimator=lr,
    param_grid={'solver':['sag','saga'],
                'C':[0.5,0.75,1.0,1.25,1.5,1.75,2.0]
            },
    cv = 15
)
```

```python
gd.fit(x,y)
#getting best values
print("best parameters ;{}".format(gd.best_params_))
print("best score:{}".format(gd.best_score_))
print("best model:{}".format(gd.best_estimator_))
```

```python
lr = LogisticRegression(C = 0.5,solver='sag')
lr.fit(x,y)
r = lr.predict(xtest)
```
