```python
data=pd.read_csv('../input/nlp-getting-started/train.csv',usecols=['text', 'target'])
```

```python
data['target'].value_counts()
```

```python
import re
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
ls=WordNetLemmatizer()
```

```python
corpus=[]
for i in range(0,len(data)):
    tweets=re.sub('[^a-zA-Z]',' ',data['text'][i])
    tweets=tweets.lower()
    tweets=tweets.split()
    tweets=[ls.lemmatize(word) for word in tweets if word not in stopwords.words('english')]
    tweets=' '.join(tweets)
    corpus.append(tweets)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
cv=TfidfVectorizer(max_features=2500)
X=cv.fit_transform(corpus).toarray()
```

```python
X
```

```python
y=data['target']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB().fit(X_train,y_train)
```

```python
y_predd=model.predict(X_test)
```

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(y_test,y_predd)
```

```python
metrics.confusion_matrix(y_test,y_predd)
```

```python
from sklearn.ensemble import RandomForestClassifier
mod=RandomForestClassifier().fit(X_train,y_train)
```

```python
y_preddd=mod.predict(X_test)
```

```python
metrics.accuracy_score(y_test,y_preddd)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression().fit(X_train,y_train)
```

```python
y_pr=lr.predict(X_test)
```

```python
metrics.accuracy_score(y_test,y_pr)
```
