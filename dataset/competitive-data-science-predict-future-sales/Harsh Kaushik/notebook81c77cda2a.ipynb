```python
features = ['shop_id','item_id']
x=df[features]
y = df.item_cnt_day
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

my_pipeline = Pipeline(steps=[('preprocessor', SimpleImputer()),
                              ('model', RandomForestRegressor(n_estimators=50,
                                                              random_state=0))
                             ])
```

```python
train_x,val_x,train_y,val_y= train_test_split(x,y,random_state=1,test_size=0.3,train_size=0.7)
```

```python
from sklearn.tree import DecisionTreeRegressor

# Define model. Specify a number for random_state to ensure same results each run
my_model = DecisionTreeRegressor(random_state=1)

# Fit model
my_model.fit(train_x, train_y)
from sklearn.metrics import mean_absolute_error

predictions = my_model.predict(val_x)
print("Mean Absolute Error: " + str(mean_absolute_error(val_y, predictions)))
# predictions2= my_model2.predict(val_x)
# print("mean ab error: "+ str(mean_absolute_error(val_y,predictions2) ))
```
