```python
sales_train = pd.read_csv('/kaggle/input/competitive-data-science-predict-future-sales/sales_train.csv')
test = pd.read_csv('/kaggle/input/competitive-data-science-predict-future-sales/test.csv')
```

```python
sales_train = sales_train.sample(frac=1).reset_index(drop=True)
sales_train = sales_train.loc[:1000000,]
sales_train = pd.get_dummies(sales_train, columns=['shop_id'])
sales_train = pd.get_dummies(sales_train, columns=['item_id'])
```

```python
X_train = np.array(sales_train.loc[:900000,].drop(['date','date_block_num','item_cnt_day'],1))
y_train = np.array(sales_train['item_cnt_day'].loc[:900000,])
X_dev = np.array(sales_train.loc[900000:950000,].drop(['date','date_block_num','item_cnt_day'],1))
y_dev = np.array(sales_train['item_cnt_day'].loc[900000:950000,])
X_test = np.array(sales_train.loc[950000:,].drop(['date','date_block_num','item_cnt_day'],1))
y_test = np.array(sales_train['item_cnt_day'].loc[950000:,])
```

```python
scaler = StandardScaler()   
X_train = scaler.fit_transform(X_train)
X_dev = scaler.fit_transform(X_dev)
X_test = scaler.fit_transform(X_test)
```

```python
model1 = Sequential()
model1.add(Dense(128,input_shape=(62,),activation='relu'))
model1.add(BatchNormalization())
model1.add(Dense(256,activation='relu'))
model1.add(BatchNormalization())
model1.add(Dense(256,activation='relu'))
model1.add(BatchNormalization())
model1.add(Dense(1,activation='linear'))
```

```python
model1.compile(loss='mse', optimizer='adam')
model1.fit(X_train,y_train,epochs=5,batch_size=128)
```

```python
model1.evaluate(X_dev,y_dev)
```

```python
model2 = Sequential()
model2.add(Dense(128,input_shape=(62,),activation='relu'))
model2.add(BatchNormalization())
model2.add(Dense(256,activation='relu'))
model2.add(BatchNormalization())
model2.add(Dense(256,activation='relu'))
model2.add(BatchNormalization())
model2.add(Dense(256,activation='relu'))
model2.add(BatchNormalization())
model2.add(Dense(256,activation='relu'))
model2.add(BatchNormalization())
model2.add(Dense(1))
```

```python
model2.compile(loss='mse', optimizer='adam')
model2.fit(X_train,y_train,epochs=10,batch_size=128)
```

```python
model2.compile(loss=MeanSquaredLogarithmicError(), optimizer='adam')
model2.fit(X_train,y_train,epochs=5,batch_size=128)
```

```python
model2.evaluate(X_dev,y_dev)
```
