```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read data
path = '../input/competitive-data-science-predict-future-sales/'
df_item_categories = pd.read_csv(path + 'item_categories.csv')
df_items = pd.read_csv(path + 'items.csv')
df_sales_train = pd.read_csv(path + 'sales_train.csv', parse_dates=[0]) # https://stackoverflow.com/questions/16852911/how-do-i-convert-strings-in-a-pandas-data-frame-to-a-date-data-type
df_shops = pd.read_csv(path + 'shops.csv')
df_test = pd.read_csv(path + 'test.csv')
```

# Item Categories
supplemental information about the items categories.

```python
# Data
df_item_categories = df_item_categories.set_index('item_category_id') ## setting item_category_id as index
print(f"Total Categories Data: {df_item_categories.shape}")
print(f"Columns {df_item_categories.columns.tolist()}")
df_item_categories.head()
```

### How many unique categories do we have ?

```python
df_item_categories.item_category_name.nunique()
```

# Items

supplemental information about the items/products.

```python
# Data
df_items = df_items.set_index('item_id') ## setting id as index
print(f"Items data {df_items.shape}")
print(f"Items columns {df_items.columns.tolist()}")
df_items.head()
```

### Which item categories have items in the store

```python
df_items.item_category_id.nunique()
```

All 84 categories have items in the store

```python
# Distribution of top 5 categories
item_category_id_count = df_items.item_category_id.value_counts()[:5]
plt.barh(df_item_categories.loc[item_category_id_count.index].item_category_name, item_category_id_count.values)
plt.ylabel("Item Category Name")
plt.xlabel("Count")
plt.plot();
```

DVD Category has the highest items/products in the store, followed by CD then Blu-Ray

# Shops
supplemental information about the shops.

```python
# Data
df_shops = df_shops.set_index('shop_id') # setting shop_id as index
print(f"Shops Data {df_shops.shape}")
print(f"Shops Columns {df_shops.columns.tolist()}")
df_shops.head()
```

# Sales
the training set. Daily historical data from January 2013 to October 2015

```python
# Data
print(f"Sales data {df_sales_train.shape}")
print(f"Sales Columns {df_sales_train.columns.tolist()}")
df_sales_train.head()
```

```python
# Add columns for year, day of month and month
df_sales_train['year'] = df_sales_train.date.dt.year
df_sales_train['month'] = df_sales_train.date.dt.month
df_sales_train['day'] = df_sales_train.date.dt.day
df_sales_train.head()
```

### How many years data is present ?

```python
df_sales_train.year.unique().tolist()
```

3 years data is present 13, 14 and 15

### How many shops do we have ?

```python
df_sales_train.shop_id.nunique()
```

Total 60 shops items we have in dataset

### Top 5 trending shops

```python
df_top_shops_id = df_sales_train.shop_id.value_counts()[:5]
plt.barh(df_shops.loc[df_top_shops_id.index].shop_name, df_top_shops_id.values)
plt.ylabel("Shop Name")
plt.xlabel("Count")
plt.plot();
```

### Top 5 trending products

```python
df_top_items_id = df_sales_train.item_id.value_counts()[:5]
plt.barh(df_items.loc[df_top_items_id.index].item_name, df_top_items_id.values)
plt.ylabel("Item Name")
plt.xlabel("Count")
plt.plot();
```

### Distribution of item price

```python
print(f"Range of item prices: {df_sales_train.item_price.max()}  {df_sales_train.item_price.min()}")
df_sales_train.item_price.describe()
```
