<h4 style="background-color:#86BC49;color :white;,font-size:12em; border-radius: 20px; text-align: center-right; padding: 20px">
    &nbsp; &nbsp; &nbsp; &nbsp; &nbsp;Step 1 : Creating a holidays lists</h4>

```python
russion_holiday_lists=[]
russion_holiday_name=[]

for day,name in sorted(holidays.Russia(years=[2013,2014,2015]).items()):
    #print(day,name)
    russion_holiday_lists.append(day)
    russion_holiday_name.append(name)
    

#how many holidays in total 3 years
print("total holidays in 3 yrs : ",len(russion_holiday_lists))
```



<h4 style="background-color:#86BC49 ;color :white;,font-size:12em; border-radius: 20px; text-align: center-right; padding: 20px">
    &nbsp; &nbsp; &nbsp; &nbsp; &nbsp;Step 2 : Creating a seperate holiday dataframe</h4>

```python

rus_hol ={
    'rus_hol':russion_holiday_lists,
    'hol_name_ru':russion_holiday_name

}
holidays_dataset=pd.DataFrame(rus_hol)
holidays_dataset['rus_hol'] = pd.to_datetime(holidays_dataset['rus_hol'], infer_datetime_format=True) 

"""#since the holidays name are in russia I will be translating it to english
from googletrans import Translator


x=holidays_dataset['hol_name_ru']
translator = Translator()#defining translator
holidays_dataset['hol_name_en'] = holidays_dataset['hol_name_ru'].apply(lambda x: translator.translate(x, dest='en').text)
#df['translated_value'] = df['value'].apply(lambda x: translator.translate(x, dest='en').text)

#changing the format to datetime 
#df = pd.to_datetime(df['rus_hol'], infer_datetime_format=True)  """

holidays_dataset.sort_index(inplace=True)
holidays_dataset
```


<h4 style="background-color:#86BC49 ;color :white;,font-size:12em; border-radius: 20px; text-align: center-right; padding: 20px">
    &nbsp; Note: At somepoint I would like to see if the sales goes up before or after holidays .<br>For instance : we usually buy christmas present before holidays as  well as after holidays due to  YEAR END SALES</h4>

```python
data['10_days_before_holiday'] = data['date'].isin(holidays_dataset['rus_hol']-pd.DateOffset(10))
data['10_days_after_holiday'] = data['date'].isin(holidays_dataset['rus_hol']+pd.DateOffset(10))
data.head()
```

```python
group = data.groupby(['date_block_num','shop_id','item_id']).agg({'item_cnt_day': ['sum']})
group.columns = ['item_cnt_month']
group.reset_index(inplace=True)
group.head()
```

```python
data = data.merge(group, on=['date_block_num','shop_id','item_id'], how='left')
target = (data['item_cnt_month']
                                .fillna(0)
                                .clip(0,20) # NB clip target here
                                .astype(np.float16))
```

```python
data = data.drop(['date',  'item_price',  'item_name', 'item_category_name', 'week_day',  'day', 'month', 'year', 'item_cnt_day','is_holiday_dich',
                 '10_days_after_holiday','10_days_before_holiday','shop_name','region_id','region_name','item_cnt_month','week','days_week'], axis=1)
data.head()
```

```python
from sklearn.model_selection import train_test_split
from lightgbm import LGBMRegressor

import sklearn
from sklearn.linear_model import LinearRegression
X_train, X_test, Y_train, Y_test = sklearn.model_selection.train_test_split(data,target, test_size=0.30, random_state=1)
```

```python

model_lgb_regressor= LGBMRegressor(max_depth=6,
                n_estimators=350,
              num_leaves=966, reg_alpha=0.04, reg_lambda=0.073,
              subsample=0.6)


model_lgb_regressor.fit(X_train,Y_train)
```

```python
print('Test set score:', model_lgb_regressor.score(X_train,Y_train))
```
