# Negative Values ( Returns ) exclude or predict ?

```python
from pyspark.sql import functions as f
```

```python
sdf_sales_train= sdf_sales_train.withColumn('item_cnt_day', 
                           f.when(sdf_sales_train['item_cnt_day'] < 0,0)
                           .otherwise(sdf_sales_train['item_cnt_day']))
sdf_sales_train.show()
```

```python
sdf_sales_train = sdf_sales_train.withColumn('item_price',
                                            f.when(sdf_sales_train['item_price'] < 0, 0)
                                            .otherwise(sdf_sales_train['item_price']))
```

```python
sdf_sales_train.where((sdf_sales_train['item_cnt_day'] < 0) | 
                     (sdf_sales_train['item_price'] < 0)).count()
```

# Type casting / data cleaning

```python
sales_data = sdf_sales_train.toPandas()
sales_data['date'] = pd.to_datetime(sales_data['date'],format = '%d.%m.%Y')
```

```python
#unique date
unique_date = sales_data['date'].unique()
unique_date.shape
```

```python
sales_data.dtypes
# sales_data.T
```

```python
dataset = sales_data.pivot_table(index = ['shop_id','item_id'],values = ['item_cnt_day'],columns = ['date_block_num'],fill_value = 0,aggfunc='sum')
dataset
```

```python
dataset.reset_index(inplace = True)
dataset.head()
```

```python
test_data = sdf_test.toPandas()
test_data.shape
```

```python
dataset = pd.merge(test_data,dataset,on = ['item_id','shop_id'],how = 'left')
```

```python
dataset.shape
```

```python
dataset.fillna(0,inplace = True)
dataset.head()
```

```python
dataset.drop(['shop_id','item_id','ID'],inplace = True, axis = 1)
dataset.head()
```

# Train Test Split

```python
number_valid = 3000

# X we will keep all columns execpt the last one 
X_train = np.expand_dims(dataset.values[:-number_valid,:-1],axis = 2)
# the last column is our label
y_train = dataset.values[:-number_valid,-1:]

X_valid = np.expand_dims(dataset.values[-number_valid:,:-1],axis = 2)
y_valid = dataset.values[-number_valid:,-1:]

# for test we keep all the columns execpt the first one
X_test = np.expand_dims(dataset.values[:,1:],axis = 2)

# lets have a look on the shape 
print(dataset.shape, X_train.shape, y_train.shape, X_valid.shape, y_valid.shape, X_test.shape)
```

# TF Model building

```python
lr_schedule = tf.keras.optimizers.schedules.ExponentialDecay(
    initial_learning_rate=0.001,
    decay_steps=30,
    decay_rate=0.96,
    staircase=False)

# lr rate plot
steps = np.arange(3000)
fig, axes = plt.subplots()
axes.plot(steps, lr_schedule(steps))

model  = tf.keras.models.Sequential()
# model.add(tf.keras.layers.LSTM(32,input_shape=X_train.shape[-2:]))
model.add(tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(32), input_shape=X_train.shape[-2:]))
model.add(tf.keras.layers.Dropout(0.4))
model.add(tf.keras.layers.Dense(1,activation='relu'))
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=lr_schedule), loss='mse', run_eagerly=False)

model.summary()

#check eager or not
model.run_eagerly
```

```python
tf.keras.utils.plot_model(model,show_layer_names=True,show_shapes=True)
```

# Train the model

```python
early_stopping = tf.keras.callbacks.EarlyStopping(patience=8, verbose=1, restore_best_weights=True)
model.fit(X_train,y_train,
          verbose = 1,
          batch_size = 4096,
          epochs = 20, 
          validation_data=(X_valid, y_valid), 
          callbacks=[early_stopping], use_multiprocessing=True)
```

```python
#check restore bast model or not 
# model.evaluate(X_valid, y_valid, batch_size=4096)
model.evaluate(X_train[:1200], y_train[:1200], batch_size=4096)
```

# Use trained model for Prediction

```python
# creating submission file 
submission_pfs = model.predict(X_test)
# creating dataframe with required columns 
submission = pd.DataFrame({'ID':test_data['ID'],'item_cnt_month':submission_pfs.ravel()})
# creating csv file from dataframe
submission.T
```
