```python
cat=pd.read_csv('/kaggle/input/competitive-data-science-predict-future-sales/item_categories.csv')
cat.head()
```

```python
def r(x):
    if 'PC' in x:
        return 'PC'
    elif 'PS2' in x:
        return 'PS2'
    elif 'PS3' in x:
        return 'PS3'
    elif 'PSP' in x:
        return 'PSP'
    elif 'PS4' in x:
        return 'PS4'
    elif 'PSVita' in x:
        return 'PSVita'
    elif 'XBOX 360' in x:
        return 'XBOX 360'
    elif 'XBOX ONE' in x:
        return 'XBOX ONE'
    elif 'Blu-Ray 3D' in x:
        return 'Blu-Ray 3D'
    elif 'Blu-Ray 4K' in x:
        return 'Blu-Ray 4K'
    
    else:
        return 'Others'
cat['item_category_name']=cat['item_category_name'].apply(r)
```
