```python
import pandas as pd
data = pd.read_csv("../input/competitive-data-science-predict-future-sales/sales_train.csv")
test = pd.read_csv("../input/competitive-data-science-predict-future-sales/test.csv")
data = data.drop(columns = ["date"])
data = data.drop(columns = ["date_block_num"])
```

```python
data.head()
```

```python
X = data.iloc[:, :3]
y = data.iloc[:, 3:]
```

```python
from sklearn.preprocessing import Normalizer
transformer = Normalizer().fit(X)
transformer.transform(X)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier(max_depth = 10, random_state = 0)
```

```python
clf.fit(X, y)
```

```python
y_pred = clf.predict(X)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y, y_pred)
```

```python
X.head()
```

```python
y_prd = clf.predict(test)
```

```python
# Create the submission file and submit
preds = pd.DataFrame(y_prd, columns=['item_cnt_month'])
preds.to_csv('submission.csv',index_label='ID')
```
