```python
sales = sales[sales.item_price<30000]
sales = sales[sales.item_cnt_day<200]
sales = sales[sales.item_cnt_day>-1]
columns = ['date', 'date_block_num', 'shop_id', 'item_id','item_price','item_cnt_day']
sales.drop_duplicates(columns,keep='first', inplace=True)
```

```python
data = sales[['item_cnt_day','item_price']]
data.head()
```

```python
x = data.iloc[:, :-1].values
y = data.iloc[:, 1].values

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 2/3, random_state = 246, shuffle=2)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

plt.scatter(X_train, y_train, color = 'blue')
plt.plot(X_train, model.predict(X_train), color = 'yellow')
plt.xlabel('item_cnt_day')
plt.ylabel('item_price')
plt.show()
```

```python
r2_score(y_test, y_pred)
```
