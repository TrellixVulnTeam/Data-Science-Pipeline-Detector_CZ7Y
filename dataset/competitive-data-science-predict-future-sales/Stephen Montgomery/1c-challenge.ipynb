#### Visualise data

```python
# List of features
cols = list(train.columns)
# Drop date column
cols.remove('date')
cols
```

```python
for i in cols:
    sns.boxplot(x=train[i])
    plt.show()
```

Looks like we have some outliers in item price and item_cnt_day. Before removing them, I want to see them in relation to the time-series:

```python
for i in cols:
    sns.lineplot(x=train.index ,y=train[i])
    plt.show()
```
