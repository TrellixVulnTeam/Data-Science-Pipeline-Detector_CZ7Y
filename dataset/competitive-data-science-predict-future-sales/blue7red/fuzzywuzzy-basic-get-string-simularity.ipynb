# fuzzywuzzy

Fuzzy string matching like a boss. 

It uses Levenshtein Distance to calculate the differences between sequences in a simple-to-use package.

```python
from fuzzywuzzy import fuzz,process
```

# 1.ratio  => ratio simularity

```python
def getSimularityScore(str1,str2):
    return fuzz.ratio(str1.lower(), str2.lower())
    

a = 'Python is programming language'
b = 'Pithon is Programing lang'

ratio = getSimularityScore(a,b)
print('Similarity score: {}'.format(ratio))
```

# 2.partial_ratio => partial ratio simularity

```python
def getSimularityPartialScore(str1,str2):
    return fuzz.partial_ratio(str1.lower(), str2.lower())


a = 'Python language' 
b = 'Python'

ratio = getSimularityPartialScore(a,b)
print('Similarity score: {}'.format(ratio))
```

# 3.token_sort_ratio = > allow sort + ratio simularity

```python
def getSimularityTokenSortScore(str1,str2):
    return fuzz.token_sort_ratio(str1.lower(), str2.lower())


a = 'apple banna orange' 
b = 'orange apple banna'

ratio = getSimularityTokenSortScore(a,b)
print('Similarity score: {}'.format(ratio))
```
