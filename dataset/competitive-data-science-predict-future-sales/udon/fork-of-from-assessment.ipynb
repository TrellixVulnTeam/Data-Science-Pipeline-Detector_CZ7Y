### 一つ目の計測方法

```python
ans = pd.merge(sub, new_test)
```

```python
ans
```

```python
ans.query("ID == 2909401 or ID == 2909818")
```

```python
clear = ans[ans["ID"] != 2909401]
clear
```

```python
clear = ans[ans["ID"] != 2909818]
clear
```

```python
clear["e"] = clear["item_cnt_month"] - clear["item_cnt_day"]
clear
```

```python
clear["e**2"] = clear["e"]**2
clear
```

```python
clear["e**2"].sum()
```

```python
total = clear["e**2"].sum() / len(clear)
total
```

```python
final_ans = np.sqrt(total)
print(final_ans)
```
