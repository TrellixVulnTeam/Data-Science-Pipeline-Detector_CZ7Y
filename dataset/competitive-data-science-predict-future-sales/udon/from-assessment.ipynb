```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
sub = pd.read_csv("../input/udon-future-predict/xgb_submission.csv")
```

```python
sub
```

```python
data = pd.read_csv('../input/competitive-data-science-predict-future-sales/sales_train.csv')
test = data[data.date_block_num == 33]
drop_cols =["date","date_block_num","shop_id","item_price","item_id"]
test.drop(drop_cols, axis=1,inplace = True)
test
```

```python
test = test.reset_index()
```

```python
new_test = test.rename(columns={'index':'ID',"item_cnt_day":"item_cnt_day"})
new_test.head()
```

```python
sub
```

### 一つ目の計測方法

```python
ans = pd.concat([sub, new_test], axis=1)
```

```python
ans
```

```python
ans["e"] = ans["item_cnt_month"] - ans["item_cnt_day"]
ans
```

```python
ans["e**2"] = ans["e"]**2
ans
```

```python
ans["e**2"].sum()
```

```python
total = ans["e**2"].sum() / len(ans)
total
```

```python
final_ans = np.sqrt(total)
print(final_ans)
```
