```python
SatisEgitimi = SatisEgitimi[SatisEgitimi.item_price<45000]
SatisEgitimi = SatisEgitimi[SatisEgitimi.item_cnt_day<600]
columns = ['date', 'date_block_num', 'shop_id', 'item_id','item_price','item_cnt_day']
SatisEgitimi.drop_duplicates(columns,keep='first', inplace=True) 
```

```python
plt.figure(figsize=(10,10))
plt.scatter(SatisEgitimi.item_cnt_day,SatisEgitimi.item_price)
plt.show()
```

```python
Örnek = Örnek[Örnek.ID<45000]
Örnek = Örnek[Örnek.item_cnt_month<600]
columns = ['ID', 'item_cnt_month']
Örnek.drop_duplicates(columns,keep='first', inplace=True) 
```

```python
plt.figure(figsize=(10,10))
plt.scatter(Örnek.ID,Örnek.item_cnt_month)
plt.show()
```

```python
monthly = SatisEgitimi.groupby(["date_block_num","shop_id","item_id"])["item_cnt_day"].agg('sum').reset_index()
monthly.columns = ['date_block_num','shop_id','item_id','item_cnt_month']

price_monthly = SatisEgitimi.groupby(["date_block_num","shop_id","item_id"])["item_price"].agg('mean').reset_index()

combine = pd.merge(monthly, price_monthly)
combine.head(5)
```
