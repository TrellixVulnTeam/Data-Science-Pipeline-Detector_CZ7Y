## 1. Using dataprep for simple EDA

```python
import pandas as pd
from dataprep.eda import plot, plot_correlation, plot_missing, create_report
```

```python
item_categories = pd.read_csv("../input/competitive-data-science-predict-future-sales/item_categories.csv")
items = pd.read_csv("../input/competitive-data-science-predict-future-sales/items.csv")
sales_train = pd.read_csv("../input/competitive-data-science-predict-future-sales/shops.csv")
```

```python
plot(items)
```

```python
plot_correlation(items)
```

```python
plot_missing(items)
```

```python
plot(sales_train)
```

```python
plot_correlation(sales_train)
```

```python
plot_missing(sales_train)
```

## 2. Save EDA report (Format : HTML)

```python
create_report(sales_train).save("./sale_train_eda")
create_report(items).save("./items_eda")
create_report(item_categories).save("./item_categories_eda")
```
