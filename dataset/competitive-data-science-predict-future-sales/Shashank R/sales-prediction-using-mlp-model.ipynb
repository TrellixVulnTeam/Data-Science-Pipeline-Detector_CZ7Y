```python
sales_train=pd.read_csv('/kaggle/input/competitive-data-science-predict-future-sales/sales_train.csv')
sales_train.head()
```

```python
sales_train.shape
```

```python
items=pd.read_csv('/kaggle/input/competitive-data-science-predict-future-sales/items.csv')
items.head()
```

```python
item_categories=pd.read_csv('/kaggle/input/competitive-data-science-predict-future-sales/item_categories.csv')
item_categories.head()
```

```python
shop=pd.read_csv('/kaggle/input/competitive-data-science-predict-future-sales/shops.csv')
shop.head()
```

```python
test=pd.read_csv('/kaggle/input/competitive-data-science-predict-future-sales/test.csv')
test.head()
```

```python
df1=pd.merge(items,item_categories,on='item_category_id',how='inner')
df2=pd.merge(sales_train,shop,on='shop_id',how='inner')
df=pd.merge(df1,df2,on='item_id',how='inner')
```

```python
df.head()
```

```python
df.shape
```

```python
df = df[df['shop_id'].isin(test['shop_id'].unique())]
df = df[df['item_id'].isin(test['shop_id'].unique())]
```

```python
df.head()
```

```python
df.shape
```

```python
df.head()
```

```python
df['date']=pd.to_datetime(df['date'])
df['day']=df['date'].dt.day
df['month']=df['date'].dt.month
df['year']=df['date'].dt.year
```

```python
df.head()
```

```python
df=df.drop(columns=['item_name','item_category_name','date','date_block_num','shop_name'])
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
X=df[['item_id','item_category_id','shop_id','item_price','day','month','year']]
y=df[['item_cnt_day']]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=1)
```

```python
mmc=MinMaxScaler()
```

```python
X_train=mmc.fit_transform(X_train)
X_test=mmc.transform(X_test)
```

```python
clf=MLPClassifier(random_state=1,max_iter=300)
```

```python
clf.fit(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

```python
y_pred=clf.predict(X_test)
```

```python
ID = df["item_id"]
results = pd.concat([ID, y_pred], axis=1)
```

```python
results.to_csv('Sales_prediction.csv',index=False)
```
