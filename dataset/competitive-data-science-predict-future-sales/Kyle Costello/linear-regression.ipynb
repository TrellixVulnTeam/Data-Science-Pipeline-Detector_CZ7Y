```python
data = pd.read_csv("../input/competitive-data-science-predict-future-sales/sales_train.csv") # load data set
```

```python
def monthly_sales(data):
    data = data.copy()     
    data = data.groupby(['shop_id', 'item_id', 'date_block_num'])['item_cnt_day'].sum()
    return data

monthly_data = monthly_sales(data)
print(monthly_data.column)
```

```python
def splitTrainTest(data):
    data = data.drop(['item_cnt_month','date'],axis=1)
    train, test = np.array(data[0:-12].values), np.array(data[-12:].values)
    
    return train, test

train, test = splitTrainTest(monthly_data)
```

```python
def scale_data(train_set, test_set):
    #apply Min Max Scaler
    #scaler = MinMaxScaler(feature_range=(-1, 1))
    #scaler = scaler.fit(train_set)
    
    #reshape training set
    train_set = train_set.reshape(train_set.reshape[0], train_set.shape[1])
    test_set = test_set.reshape(test_set.reshape[0], test_set.shape[1])
    return train_set, test_set

train, test = scale_data(train, test)
```
