```python
#importing the package and loading the dataset
import pandas as pd 
import numpy as  np 
sales   =  pd.read_csv("../input/competitive-data-science-predict-future-sales/sales_train.csv")
```

```python
sales.head()
```

```python
#Checking description
sales.describe()
```

```python
#Checking for missing values
sales.isnull().sum()
```

Since the Test series requires the predictionthought 3 parameters date , shop_id and item_id So droping date_block_num & item_price

```python
#dropping unecessary columns
sales.drop(['date_block_num','item_price'],axis = 1 ,inplace = True)
```

Applying proper formating to the date to extract year and month of each sale

```python
#Applying format to get day first in datetime sequence
sales['date']   =   pd.to_datetime(sales['date'],dayfirst=True)

#Extracting months yearwise from date
sales['date']   =   sales['date'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
sales["date"]
```

Creating pivot table takeing each month as a column in dataframe

```python
sales.head()
```

```python
sales   =   sales.groupby(['date','shop_id','item_id']).sum()
sales   =   sales.pivot_table(index=['shop_id','item_id'],columns = 'date',values = 'item_cnt_day',fill_value = 0)
sales.reset_index(inplace = True)




sales.head()
```

Splitting the dependent and independent variable for training our  data 

```python
y_train = sales['2015-10'].values
x_train = sales.drop(['2015-10'],axis=1)
```

fiiting out data to Randomforest model

```python
#importing the package and fitting the data 
from sklearn.ensemble import RandomForestClassifier
rf  =  RandomForestClassifier(n_estimators=100)
rf.fit(x_train,y_train)
```

importing and the test data and filling the dependent columns for prediction

```python
sales_test  =   pd.read_csv("../input/competitive-data-science-predict-future-sales/test.csv")
```

```python
sales_test = pd.merge(sales_test, sales, on=['shop_id','item_id'], how='left')
sales_test.drop(['ID', '2013-01'], axis=1, inplace=True)
sales_test = sales_test.fillna(0)
sales_test.head()
```

prediction....

```python
predictions  =   rf.predict(sales_test)
#formatting 
predictions =   list(map(round,predictions))
```

Creating final result as given by sample submission

```python
final = pd.DataFrame({'ID':np.arange(0,len(predictions),1),'item_cnt_month':predictions})
final.head()
```

```python
final.to_csv('submission.csv',index=False,encoding='utf-8')
```
