#### **Changing date column's type to datetime type**

```python
#7 CHANGING date DTYPE

import datetime as dt

sales_train['date'] = pd.to_datetime(sales_train['date'], format='%d.%m.%Y')

sales_train['day'] = sales_train['date'].dt.day
sales_train['month'] = sales_train['date'].dt.month
sales_train['year'] = sales_train['date'].dt.year

sales_train.drop('date', axis=1, inplace=True)

sales_train
```

```python
#8 THERE ARE VALUES SMALLER THAN ZERO

sales_train[sales_train['item_cnt_day'] < 0]
```

```python
#9 I CHANGE THESE VALUES FROM NEGATIVE TO POSITIVE

sales_train['item_cnt_day'] = sales_train['item_cnt_day'].map(lambda x: np.abs(x))
sales_train
```

# **Preparation**

```python
#13 COPYING from sales_train

df = sales_train.copy()
df
```

```python
#14 DROPPING SOME COLUMNS

df.drop(['date_block_num', 'item_price'], axis=1, inplace=True)
df
```

```python
#15 SORTING COLUMNS | X : y

df = df[['shop_id', 'item_id', 'day', 'month', 'year', 'item_cnt_day']]
df
```

```python
#16 TRAIN_TEST_SPLIT => train_set, test_set

from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(df, test_size=0.2, random_state=29)
```

```python
#17 train_set => X_train, y

X_train = train_set.drop('item_cnt_day', axis=1)
y = train_set['item_cnt_day'].copy()
```

```python
#18 X_train => MinMaxScaler

from sklearn.preprocessing import MinMaxScaler

min_max_scaler = MinMaxScaler()

X_prepared = min_max_scaler.fit_transform(X_train)
```

# **Machine Learning** **section**

# **Linear Regression**

```python
#19 LINEAR REGRESSION

from sklearn.linear_model import LinearRegression

LR_model = LinearRegression()
LR_model.fit(X_prepared, y)
```

# **LR - Prediction test**

```python
#20 TEST_SET
test_set
```

```python
#21 test_set => X_test, y_test

X_test = test_set.drop('item_cnt_day', axis=1)
y_test = test_set['item_cnt_day'].copy()
```

```python
#22 X_test => X_test_prepared

X_test_prepared = min_max_scaler.transform(X_test)
```

```python
#23 LR_model prediction

y_predicted = LR_model.predict(X_test_prepared)

print("Predictions:")
for i in y_predicted[:10]:
    print(i)
    
print("Orginal test:")
y_test.iloc[:10]
```

# **LR - MeanAbsoluteError**

```python
#24 LR_model => mae

from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(y_test, y_predicted)

print("MAE=", mae)
```

# **LR - MeanSquaredError**

```python
#25 => mse

from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, y_predicted)

print("RMSE=", np.sqrt(mse))
```

```python
#35 FUNCTION

def display_scores(scores):
   print("Scores:", scores)
   print("Mean:", scores.mean())
   print("Std.dev:", scores.std())
```

```python
scores = cross_val_score(RF_model, Xcross_prepared, ycross, scoring="neg_mean_squared_error", cv=3)
RF_rmse_scores = np.sqrt(-scores)
display_scores(RF_rmse_scores)
```
