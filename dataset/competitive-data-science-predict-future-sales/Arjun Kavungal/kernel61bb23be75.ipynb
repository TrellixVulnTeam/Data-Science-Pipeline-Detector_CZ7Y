```python
test = pd.read_csv('/kaggle/input/competitive-data-science-predict-future-sales/test.csv')
test
```

```python
items = pd.read_csv("/kaggle/input/competitive-data-science-predict-future-sales/items.csv")
```

```python
train = pd.read_csv('/kaggle/input/competitive-data-science-predict-future-sales/sales_train.csv')
train = train.merge(items, on="item_id")
train["item_category_id"] = train["item_category_id"].astype("category")
```

```python
train.columns
test.columns
```

```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
X = train[['shop_id','item_id']]
y = train[['item_cnt_day']]
model = DecisionTreeRegressor().fit(X,y)
p = model.predict(test[['shop_id','item_id']])
f = pd.DataFrame(p)
f = f.reset_index()
f = f.rename(columns={"index": "ID", 0: "item_cnt_month"})
f.to_csv('s2.csv')
f
```
