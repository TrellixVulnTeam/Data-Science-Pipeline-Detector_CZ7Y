```python
sales_train = pd.read_csv('/kaggle/input/competitive-data-science-predict-future-sales/sales_train.csv')
test = pd.read_csv('/kaggle/input/competitive-data-science-predict-future-sales/test.csv')
```

```python
sales_train.head()
```

(date_block_num','shop_id','item_id','item_cnt_day) verilerini (date_block_num','shop_id','item_id) verilerine göre grupluyoruz ve ardından aylara göre satışı bulmak için sum komutu ile topluyoruz

```python
df = sales_train[['date_block_num','shop_id','item_id','item_cnt_day']].groupby(['date_block_num','shop_id','item_id']).sum().reset_index()
df
```

X ve Y değerlerimizi alıyoruz . X değerleri = Daha önceden ayırdığımız aylar , mağazalar, ürünler .Y değeri ise satış miktarı 

```python
y = df.item_cnt_day.values

x = df.drop(["item_cnt_day"], axis=1)
```

Naive Bayes Teoremi ile normalizasyon işlemi

```python
x = (x - np.min(x)) / (np.max(x) - np.min(x))

x
```

Reshape işlemi ile verilerimize tekrar şekillendirerek model için uygun hale getiriyoruz.

```python
y = y.reshape(-1,1)

y.shape
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

Naive Bayes uygulanması

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(x, y)
```

Model doğruluk kontrolü

```python
print(nb.score(x_test, y_test))
```

```python
y_head = nb.predict(x_test)
```

```python
tbl = x_test
tbl['results'] = pd.Series(y_head)
tbl.describe()
```
