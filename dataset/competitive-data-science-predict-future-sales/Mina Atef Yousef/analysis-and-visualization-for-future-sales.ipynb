# Importind and Cleaning data

## Import the required Lib and data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
items= pd.read_csv('/kaggle/input/competitive-data-science-predict-future-sales/items.csv')
item_categories=pd.read_csv('/kaggle/input/competitive-data-science-predict-future-sales/item_categories.csv')
sales_train=pd.read_csv('/kaggle/input/competitive-data-science-predict-future-sales/sales_train.csv')                            
shops=pd.read_csv('/kaggle/input/competitive-data-science-predict-future-sales/shops.csv')
test=pd.read_csv('/kaggle/input/competitive-data-science-predict-future-sales/test.csv')
```

### Concatenate the data

```python
df_train=sales_train.merge(items, on='item_id')
df_train=df_train.merge(item_categories,on='item_category_id')
df_train=df_train.merge(shops,on='shop_id')
```

```python
df_train.describe()
```

```python
print("The duplicated data is: ",df_train.duplicated().sum())
print("The duplicated data is: ",df_train.isnull().sum())
```

Drop all of duplicated

```python
df_train.drop_duplicates(inplace=True)
```

```python
df_train.head()
```

```python
df_train.info()
```

```python
#change the time format
df_train.date=pd.to_datetime(df_train.date)
```

```python
df_train[df_train['date_block_num']==0]
```

```python
gp=df_train.groupby('shop_name')['item_cnt_day'].count().reset_index(name='count').sort_values(['count'], ascending=False).head(5)
sns.barplot(data=gp,x='shop_name',y='count')
_=plt.title('The top 5 of places sold the product')
_=plt.xlabel('Shop name')
_=plt.xticks(rotation=90)

plt.show()
```

The item is increased

```python
sns.barplot(data=df_train,x='date_block_num',y='item_cnt_day')
_=plt.title('The count of sold produce according to the month number')
_=plt.xlabel('month number')
_=plt.xticks(rotation=90)

plt.show()
```

```python
sns.lineplot(x='date',y="item_cnt_day",data=df_train)
plt.xlabel("The date")
plt.ylabel("item count day")
plt.xticks(rotation=90)

plt.title("The realtion between the date and the item count day")
```

```python
sns.boxplot(x='item_price',data=df_train)
```

```python
print('the mean of item price is '+ str(df_train.item_price.mean())+ 'and the std is '+str(df_train.item_price.std()))
```

```python
#remove the outlier in the price of the items, could effect on the predicitions on the future
df_price_test=df_train.drop(df_train[df_train.item_price>df_train.item_price.mean()+df_train.item_price.std()*3].index)
```
