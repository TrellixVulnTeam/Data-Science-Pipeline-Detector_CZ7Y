### based on relationship between tables lets merge the dataset 


```python
items_new = items.merge(item_categories , on='item_category_id')
print(items_new.shape , items.shape)
items_new.head()
```

```python
train_new = train.merge(shops , on='shop_id')
print(train_new.shape , train.shape)
train_new = train_new.merge(items_new , on = 'item_id')
print(train_new.shape)
train_new.head()
```

```python
test_new = test.merge(shops , on='shop_id')
print(test_new.shape , test.shape)

test_new = test_new.merge(items_new , on = 'item_id')
print(test_new.shape)
test_new.head()
```

## EDA -->
#### Feature Type Identification
Lets find numerical and categorical features 

```python
numerical = [features for features in train_new.columns if train_new[features].dtypes != 'O']
categorical = [features for features in train_new.columns if train_new[features].dtypes == 'O']
print('Numerical features are : \n' + str(numerical))
print('Categorical features are : \n' + str(categorical))
```

Numerical Data Analysis -->


```python
contineous = [features for features in numerical if train_new[features].dtypes == 'float64']
contineous
```

```python
Discrete = [features for features in numerical if train_new[features].dtypes == 'int64']
Discrete
```

```python
## PLot Histogram for all Numerical features
def plt_hist(feature_lst):
    for feature in feature_lst:
        plt.hist(train_new[feature])
        plt.xlabel(feature)
        plt.ylabel("Count of" + feature)
        plt.title(feature)
        plt.show()
```

```python
plt_hist(Discrete)
```

```python
#### PLot the trend chart for the item count - Jan2013 to October 2015
fig = train_new.groupby(['date_block_num'])['item_cnt_day'].sum().plot(
    figsize=(14, 8), linewidth=2)

fig.set_title('item_cnt_day in time')
fig.set_ylabel('item_cnt_day')
plt.xticks(np.arange(0, 35, 1))
plt.show()
```

### Observations
1. Dec 2013 , Dec2014 sale goes very high

```python
#### PLot the trend chart for the item count - Jan2013 to October 2015
fig = train_new.groupby(['date_block_num' , 'shop_name'])['item_cnt_day'].sum().unstack().plot(
    figsize=(14, 8), linewidth=2)

fig.set_title('item_cnt_day in time')
fig.set_ylabel('item_cnt_day')
plt.xticks(np.arange(0, 35, 1))
plt.show() 
```
