```python
with open('../input/static-ds/submission (2).csv', 'r') as f:
    lines = f.read().split('\n')
    out_lines = []
    prev_drone = None
    for i, line in enumerate(lines):
        cur_drone = line.split()[0]
        if cur_drone != prev_drone and i > 4500:
            break
        out_lines.append(line)
        prev_drone=cur_drone

with open('submission.csv', 'w') as f:
    f.write(str(len(out_lines[1:])) + '\n')
    for line in out_lines[1:]:
        f.write(line + '\n')
```
