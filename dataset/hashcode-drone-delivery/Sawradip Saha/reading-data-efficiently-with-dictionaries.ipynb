As the first line contains necessary parameters, lets read that line fast. An assign the values to variables.

```python

f = open("/kaggle/input/hashcode-drone-delivery/busy_day.in", "r")

parameters = list(map(int,f.readline().split()))

rows = parameters[0]
cols = parameters[1]
num_drones = parameters[2]
max_turns = parameters[3]
max_weight = parameters[4]

```

So, we got...

```python
print(f" rows -> {rows}")
print(f" cols -> {cols}")
print(f" num_drones -> {num_drones}")
print(f" max_turns -> {max_turns}")
print(f" max_weight -> {max_weight}")
```

Now, for the next step I choose to use dictionaries to store data, instead of list. It turns out that looking up items in a Python dictionary is much faster than looking up items in a Python list(often a thousand times  faster, because the intention of dictionary data structure , was to access values efficiently through keys).

We will mainly have three dictionaries. For products, warehouse inventories, and ordered items. Later two, will have nested dictionaries in them.


Weight of each product(values) can be accessed from the products dictionary, using product IDs(Key)

```python
#Products dictionary

num_products = int(f.readline())

products = dict()
weights_list = list(map(int,f.readline().split()))
for product_no, product_weight in enumerate(weights_list):
    products[product_no] = product_weight

                                                 
print(products)

```

The warehouse inventory dictionary will have two keys, co-ordinates of warehouse can be acessed through "location" key. And the key "inventory" will access a dictionary where keys are the product IDs, and values are how many samples of that product is stored in that warehouse.

```python
#warehouse inventory dictionary

num_warehouse = int(f.readline())
warehouses = dict()

for warehouse_no in range(num_warehouse):

    warehouse = dict()
    warehouse["location"] = list(map(int,f.readline().split())) 
    inventory = dict()
    inventory_list = list(map(int,f.readline().split()))
    for inventory_serial, num_inventory_product in enumerate(inventory_list):
        inventory[inventory_serial] = num_inventory_product

    warehouse["inventory"] = inventory
    
    warehouses[warehouse_no] = warehouse
    
print(warehouses)
```

This is how to access the number of product type 8, in warehouse 5.

```python
%%time
print(warehouses[4]['inventory'][7])
```

The orders dictionary contains the details of of all the orders .The key "location" provides where to deliver, and  

```python

num_orders = int(f.readline())
orders = dict()

for order_no in range(num_orders):

    order = dict()
    order["location"] = list(map(int,f.readline().split())) 
    ordered_items = dict()
    num_ordered_items = int(f.readline())
    ordered_items_list = list(map(int,f.readline().split()))
    for ordered_items_serial,ordered_items_type  in enumerate(ordered_items_list):
        ordered_items[ordered_items_serial] = ordered_items_type

    order["items"] = ordered_items
    
    orders[order_no] = order
    
print(orders)
```

This is how to get the location , where to deliver the 1000th order.

```python
print(orders[999]["location"])
```
