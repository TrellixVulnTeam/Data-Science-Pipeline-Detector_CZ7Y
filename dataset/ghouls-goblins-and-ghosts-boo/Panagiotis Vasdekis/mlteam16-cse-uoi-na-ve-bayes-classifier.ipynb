```python
train_data=pd.read_csv('../input/ghouls-goblins-and-ghosts-boo/train.csv.zip', index_col=0)
print('Train data: ', train_data.shape, '\n')
print(train_data.head(), '\n')

test_data=pd.read_csv('../input/ghouls-goblins-and-ghosts-boo/test.csv.zip')
test_id=test_data.values[:,0]
test_data.set_index('id', inplace=True)
print('Test data: ', test_data.shape, '\n')
print(test_data.head(), '\n')
```

```python
train_data.replace('clear', 1/6, inplace=True)
train_data.replace('green', 2/6, inplace=True)
train_data.replace('black', 3/6, inplace=True)
train_data.replace('white', 4/6, inplace=True)
train_data.replace('blue', 5/6, inplace=True)
train_data.replace('blood', 6/6, inplace=True)
print(train_data, '\n')

test_data.replace('clear', 1/6, inplace=True)
test_data.replace('green', 2/6, inplace=True)
test_data.replace('black', 3/6, inplace=True)
test_data.replace('white', 4/6, inplace=True)
test_data.replace('blue', 5/6, inplace=True)
test_data.replace('blood', 6/6, inplace=True)
print(test_data, '\n')
```

```python
X_train_Gaussian = train_data.iloc[:,:-2].values
y_train = train_data.iloc[:, -1].values
print(X_train_Gaussian.shape)
print(y_train.shape, '\n')

X_train_Multinomial = train_data.iloc[:,-2].values
X_train_Multinomial = X_train_Multinomial.reshape(-1, 1)
print(X_train_Multinomial.shape, '\n')

X_test_Gaussian = test_data.iloc[:,:4].values
print(X_test_Gaussian.shape, '\n')

X_test_Multinomial = test_data.iloc[:, -1].values
X_test_Multinomial = X_test_Multinomial.reshape(-1, 1)
print(X_test_Multinomial.shape)
```

```python
# [Îœethod 4] Naive Bayes classifier

from sklearn.naive_bayes import GaussianNB
from sklearn import metrics

nb_Gaussian = GaussianNB()
nb_Gaussian.fit(X_train_Gaussian, y_train)

y_pred_Gaussian = nb_Gaussian.predict(X_test_Gaussian)
print(y_pred_Gaussian.shape)

y_pred_f1_Gaussian = nb_Gaussian.predict(X_train_Gaussian)
print(y_pred_f1_Gaussian.shape)
y_f1_Gaussian = metrics.f1_score(y_train, y_pred_f1_Gaussian, average='weighted')
print('\nF1 score (from training set, normal distribution) = ', y_f1_Gaussian)


from sklearn.naive_bayes import MultinomialNB

nb_Multinomial = MultinomialNB()
nb_Multinomial.fit(X_train_Multinomial, y_train)

y_pred_Multinomial = nb_Multinomial.predict(X_test_Multinomial)
print(y_pred_Multinomial.shape)

y_pred_f1_Multinomial = nb_Multinomial.predict(X_train_Multinomial)
print(y_pred_f1_Multinomial.shape)
y_f1_Multinomial = metrics.f1_score(y_train, y_pred_f1_Multinomial, average='weighted')
print('\nF1 score (from training set, multinomial distribution) = ', y_f1_Multinomial)
```

```python
submission = pd.DataFrame({'id': test_id, 'type': y_pred_Gaussian})
#submission = pd.DataFrame({'id': test_id, 'type': y_pred_Multinomial})

submission.to_csv("submission.csv", index=False)
```
