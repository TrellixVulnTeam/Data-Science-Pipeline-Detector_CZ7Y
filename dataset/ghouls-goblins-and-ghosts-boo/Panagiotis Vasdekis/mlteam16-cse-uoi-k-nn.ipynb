```python
train_data=pd.read_csv('../input/ghouls-goblins-and-ghosts-boo/train.csv.zip', index_col=0)
print('Train data: ', train_data.shape, '\n')
print(train_data.head(), '\n')

test_data=pd.read_csv('../input/ghouls-goblins-and-ghosts-boo/test.csv.zip')
test_id=test_data.values[:,0]
test_data.set_index('id', inplace=True)
print('Test data: ', test_data.shape, '\n')
print(test_data.head(), '\n')
```

```python
train_data.replace('clear', 1/6, inplace=True)
train_data.replace('green', 2/6, inplace=True)
train_data.replace('black', 3/6, inplace=True)
train_data.replace('white', 4/6, inplace=True)
train_data.replace('blue', 5/6, inplace=True)
train_data.replace('blood', 6/6, inplace=True)
print(train_data, '\n')

test_data.replace('clear', 1/6, inplace=True)
test_data.replace('green', 2/6, inplace=True)
test_data.replace('black', 3/6, inplace=True)
test_data.replace('white', 4/6, inplace=True)
test_data.replace('blue', 5/6, inplace=True)
test_data.replace('blood', 6/6, inplace=True)
print(test_data, '\n')
```

```python
X_train = train_data.iloc[:,:-1].values
y_train = train_data.iloc[:, -1].values
print(X_train.shape)
print(y_train.shape, '\n')

X_test = test_data.iloc[:,:5].values
print(X_test.shape)
```

```python
# [Îœethod 1] k-NN Nearest Neighbor

from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

#classifier = KNeighborsClassifier(n_neighbors=1, metric='euclidean')
#classifier = KNeighborsClassifier(n_neighbors=3, metric='euclidean')
#classifier = KNeighborsClassifier(n_neighbors=5, metric='euclidean')
classifier = KNeighborsClassifier(n_neighbors=10, metric='euclidean')

classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)
print(y_pred.shape)

y_pred_f1 = classifier.predict(X_train)
print(y_pred_f1.shape)
y_f1 = metrics.f1_score(y_train, y_pred_f1, average='weighted')
print('\nF1 score (from training set) = ', y_f1)
```

```python
submission = pd.DataFrame({'id': test_id, 'type': y_pred})
submission.to_csv("submission.csv", index=False)
```
