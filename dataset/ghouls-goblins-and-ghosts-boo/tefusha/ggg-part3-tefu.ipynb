```python
training = pd.get_dummies(train, columns=['type'])
training
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
x1 = ['bone_length', 'rotting_flesh', 'hair_length', 'has_soul']
y1 = ['type_Ghost']
X1 = training[x1]
Y1 = training[y1]
# Build a logreg and compute the feature importances
modelghost = LogisticRegression()
# create the RFE model and select 8 attributes
modelghost.fit(X1,Y1)
```

```python
x2 = ['bone_length', 'rotting_flesh', 'hair_length', 'has_soul']
y2 = ['type_Ghoul']
X2 = training[x2]
Y2 = training[y2]
# Build a logreg and compute the feature importances
modelghoul = LogisticRegression()
# create the RFE model and select 8 attributes
modelghoul.fit(X2,Y2)
```

```python
x3 = ['bone_length', 'rotting_flesh', 'hair_length', 'has_soul']
y3 = ['type_Goblin']
X3 = training[x3]
Y3 = training[y3]
# Build a logreg and compute the feature importances
modelgoblin = LogisticRegression()
# create the RFE model and select 8 attributes
modelgoblin.fit(X3,Y3)
```

```python
from sklearn.metrics import accuracy_score
train_predicted = modelghost.predict(X1)
accuracy_score(train_predicted, Y1)
```

```python
from sklearn.metrics import accuracy_score
train_predicted = modelghoul.predict(X2)
accuracy_score(train_predicted, Y2)
```

```python
from sklearn.metrics import accuracy_score
train_predicted = modelgoblin.predict(X3)
accuracy_score(train_predicted, Y3)
```

```python
test = pd.read_csv('test.csv')
```

```python
test.isnull().sum()
```

```python
X4 = ['bone_length', 'rotting_flesh', 'hair_length', 'has_soul'] 
X4_test=test[X4]
Ghost_rate = modelghost.predict_proba(X4_test)
# print(len(Ghost_rate))
Ghoul_rate = modelghoul.predict_proba(X4_test)
# print(len(Ghoul_rate))
Goblin_rate = modelgoblin.predict_proba(X4_test)
# print(len(Goblin_rate))
type = []
for m in range(len(Ghost_rate)):
    l = [Ghost_rate[m][1],Ghoul_rate[m][1],Goblin_rate[m][1]]
    max2 = l.index(max(l))
    if max2 == 0:
        type.append('Ghost')
    elif max2 == 1:
        type.append('Ghoul')
    else:
        type.append('Goblin')
#     print(max2)
#     print(Ghost_rate[m][0])
#     print(Ghoul_rate[m][0])
#     print(Goblin_rate[m][0])
print(type)
```
