train has 371 rows and 6 columns, test has 529 rows and 5 columns

```python
train_y = train['type']
train_x = pd.DataFrame(train.drop('type', axis=1))
```

merged dataset so it more easy to preprocessing

```python
train_test=pd.concat([train_x,test])
```

```python
train_test.head()
```

color is categorical data, make it to dummis

```python
to_dummies_col = ['color']
```

make new dataframe that replace [color] into dummies

```python
train_test_dum = pd.get_dummies(train_test, columns=to_dummies_col)
```

```python
train_test_dum.head()
```

now we already have new dataframe with dummies color

split train_test_dummies into train and test

```python
train_x = train_test_dum[:len(train)]
test_x = train_test_dum[len(train):]
```

fit the features data and label (type) which is train_y

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

logreg = LogisticRegression()

import numpy as np
clf = logreg
clf.fit(train_x, np.ravel(train_y))
score = cross_val_score(clf, train_x, np.ravel(train_y), cv=5, scoring = 'accuracy')
```

```python
np.mean(score)
predict=clf.predict(test_x)
```

make new dataframe to submission and put predict into submission

```python
submission = pd.DataFrame()
submission['id']=test['id']
submission['type']=predict
```
