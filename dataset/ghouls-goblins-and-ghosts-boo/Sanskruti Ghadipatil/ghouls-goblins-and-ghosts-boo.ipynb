```python
test_data=pd.read_csv('/kaggle/input/ghouls-goblins-and-ghosts-boo/test.csv.zip')
test_data.head()
```

```python
train_data.drop('id',inplace=True,axis=1)
test_data.drop('id',inplace=True,axis=1)
```

```python
train_data.head()
```

```python
color_train=pd.get_dummies(train_data['color'])
```

```python
train_data=pd.concat([train_data,color_train],axis=1)
```

```python
train_data.head()
```

```python
train_data.drop('color',inplace=True,axis=1)
```

```python
X=train_data.drop(['type'],axis=1)
```

```python
color_test=pd.get_dummies(test_data['color'])
test_data=pd.concat([test_data,color_test],axis=1)
test_data.drop('color',inplace=True,axis=1)
test_data.head()
```

```python
y=pd.get_dummies(train_data['type'])
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test1,y_train,y_test1=train_test_split(X,y,test_size=0.1,random_state=42)
print(X_train.shape,y_train.shape)
print(X_test1.shape,y_test1.shape)
```

```python
from tensorflow import keras
from keras.layers import Dense,Dropout
from keras.models import Sequential
```

```python
model=Sequential()
model.add(Dense(64,input_shape=(X.shape[1],)))
model.add(Dense(32,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(16,activation='relu'))
model.add(Dense(8,activation='relu'))
model.add(Dense(3,activation='softmax'))
model.summary()
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
train=model.fit(X_train,y_train, 
         batch_size=16,    #batch_size-how many training exaples to use before updating weights
         #nb_epoch=20,
         epochs=10,
         verbose=2,      #verbose- determines how detailed the output will be
         validation_data=(X_test1,y_test1))
```

```python
pred=model.predict(test_data)
```
