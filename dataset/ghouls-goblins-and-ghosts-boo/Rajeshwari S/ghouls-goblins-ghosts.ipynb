```python
train_data=pd.concat([train_data,pd.get_dummies(train_data['color'])],axis=1)
train_data.drop('color',axis=1,inplace=True)
train_data.head()
```

```python
test_data = pd.concat([test_data, pd.get_dummies(test_data['color'])],axis=1)
test_data.drop('color',axis=1,inplace=True)
test_data.head()
```

```python
X=train_data.drop(['id','type'],axis=1)
y=pd.get_dummies(train_data['type'])
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=42)

print(X_train.shape,y_train.shape)
print(X_test.shape,y_test.shape)
```

```python
from tensorflow import keras
from keras.layers import Dense,Activation,Dropout
from keras.models import Sequential
```

```python
model=Sequential()

model.add(Dense(64,input_shape=(X.shape[1],)))

model.add(Dense(32,activation='relu'))

model.add(Dense(32,activation='relu'))

model.add(Dense(32,activation='relu'))

model.add(Dense(16,activation='relu'))

model.add(Dense(8,activation='relu'))

model.add(Dense(3,activation='softmax'))

model.summary()


# model=Sequential()

# model.add(Dense(64,input_shape = (X.shape[1],)) )
# model.add(Activation('relu'))

# model.add(Dense(32)
# model.add(Activation('relu'))

# model.add(Dense(32))
# model.add(Activation('relu'))

# model.add(Dense(32))
# model.add(Activation('relu'))

# model.add(Dense(16))
# model.add(Activation('relu'))

# model.add(Dense(8))
# model.add(Activation('relu'))

# model.add(Dense(3))
# model.add(Activation('softmax'))

# model.summary()
```

```python
model.compile(optimizer='adam',
             loss='categorical_crossentropy',
             metrics=['accuracy'])
```

```python
train = model.fit(X_train, y_train,
                 batch_size = 16,
                 epochs = 8,
                 verbose = 2,
                 validation_data = (X_test,y_test))
```

```python
plt.figure(figsize=(5,5))

plt.plot(train.history['accuracy'], 'r', label = 'Training accuracy')
plt.plot(train.history['val_accuracy'], 'b', label = 'Validation accuracy')

plt.legend()
plt.show()
```

```python
predictions = model.predict(test_data.drop('id', axis=1))
```

```python
pred_final=[np.argmax(i) for i in predictions]
```

```python
Submission = pd.DataFrame({'id':test_data['id'], 
                           'type':pred_final})
Submission.head()
```
