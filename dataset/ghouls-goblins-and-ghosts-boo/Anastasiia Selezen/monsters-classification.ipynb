```python
X = train_data.iloc[:,:-1].values
y = train_data.iloc[:,-1].values
```

```python
print(X.shape)
print(y.shape)
```

```python
#Splitting the data into train and test sets 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, stratify=y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.1, stratify=y_train)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
print(X_val.shape)
print(y_val.shape)
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoderX = LabelEncoder()
X_train[:, 4] = labelencoderX.fit_transform(X_train[:, 4])
X_test[:, 4] = labelencoderX.transform(X_test[:, 4])
X_val[:, 4] = labelencoderX.transform(X_val[:, 4])
```

```python
from sklearn.preprocessing import OneHotEncoder
onehotencoder = OneHotEncoder()
X_train_onehot = onehotencoder.fit_transform(X_train[:, 4].reshape(-1, 1)).toarray()
X_test_onehot = onehotencoder.transform(X_test[:, 4].reshape(-1, 1)).toarray()
X_val_onehot = onehotencoder.transform(X_val[:, 4].reshape(-1, 1)).toarray()
```

```python
X_train_onehot = X_train_onehot[:,1:]
X_test_onehot = X_test_onehot[:,1:]
X_val_onehot = X_val_onehot[:,1:]
```

```python
X_train = np.concatenate((X_train, X_train_onehot), axis=1)
X_test = np.concatenate((X_test, X_test_onehot), axis=1)
X_val = np.concatenate((X_val, X_val_onehot), axis=1)
```

```python
X_train = np.delete(X_train, 4, 1)
X_test = np.delete(X_test, 4, 1)
X_val = np.delete(X_val, 4, 1)
```

```python
print(X_train.shape)
print(X_test.shape)
print(X_val.shape)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
standardscaler = StandardScaler()
X_train = standardscaler.fit_transform(X_train)
X_test = standardscaler.transform(X_test)
X_val = standardscaler.transform(X_val)
```

```python
X_train
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoderY = LabelEncoder()
y_train = labelencoderY.fit_transform(y_train)
y_test = labelencoderY.transform(y_test)
y_val = labelencoderY.transform(y_val)
```

```python
y_train
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(C=10,max_iter=10,multi_class='multinomial',penalty='l2',solver='newton-cg', class_weight='balanced')
lr.fit(X_train,y_train)
predictions = lr.predict(X_test)
print(classification_report(y_test,predictions))
```

```python
predictions = lr.predict(X_val)
print(classification_report(y_val,predictions))
```
