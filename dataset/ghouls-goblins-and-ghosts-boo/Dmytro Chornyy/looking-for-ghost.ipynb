***DATA exploration***

```python
import numpy as np 
import pandas as pd 
import matplotlib as plt 
from sklearn import neighbors, model_selection
import seaborn as sns 

%matplotlib inline
```

Load data

```python
test_set = pd.read_csv(r"test.csv", index_col="id")
train_set = pd.read_csv(r"train.csv", index_col="id")
```

```python
print("Test shape: ", test_set.shape)
print("Train shape: ", train_set.shape)
```

```python
train_set.head(5)
```

```python
train_set.type.value_counts()
```

Data preparation

```python
X_replace_features = {
    "color":{
        "white":1,
        "clear":2,
        "green":3,
        "black":4,
        "blue": 5,
        "blood":6
    }
}
test_set = test_set.replace(X_replace_features)
Y = train_set.type
X = train_set[train_set.columns[:-1]].replace(X_replace_features)
X = (X - X.mean())/X.std()
test_set = (test_set -  test_set.mean())/test_set.std()
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
print("X_train shape: ", X_train.shape)
print("y_train shape: ", X_test.shape)
print("X_test shape: ", y_train.shape)
print("y_test shape: ", y_test.shape)
```

**Model training**

```python
clf = neighbors.KNeighborsClassifier(n_neighbors = 87, weights="distance")
clf = clf.fit(X_train, y_train)
(clf.predict(X_test) == y_test).sum()/len(y_test)
```

```python
answ = clf.predict(test_set)
```

```python
pd.DataFrame({"id": np.asarray(test_set.index), "type": answ}).to_csv("ansv.csv", index=False)
```
