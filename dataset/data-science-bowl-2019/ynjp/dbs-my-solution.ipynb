## Feature engineering

### basic
- counting: activities, game, clip, Assessment, event-code, event-id, title, 
- duration: game, clip, ativity, not study
- time-spent: each user activities
- accumulated data: accuracy group, accuracy, correct/uncorrect attempts, actions, not study time, spent hour in a day, game miss
- last application data: accuracy_title, game_title, true/false_attempts_title, activity, duration of each data
- others: not study time, game data(game round, level)
### more
- add mean, std, max of all duration data

## Selecting feature
- delete useless data(all 0 along with axis = 1 or 0)
- delete the features whose mean of train and test data have huge error

## Bayes Optimization
- apply for LGB model
- used for learning rate, feature fraction, L1/L2 normalization, max_depth, subsample_frequency
