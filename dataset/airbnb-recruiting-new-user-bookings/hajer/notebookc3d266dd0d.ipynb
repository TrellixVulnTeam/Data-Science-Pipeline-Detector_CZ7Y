```python
zip_file = "data/data.zip"
zip_dir  = zip_file.split("/")[0] # = data
    
if not os.path.exists(zip_dir):
    os.makedirs(zip_dir)
    
print ("Downloading file: " + zip_file + " ...")
urllib.request.urlretrieve('https://raw.githubusercontent.com/dataventureutc/Kaggle-HandsOnLab/master/' + zip_file, zip_file)
print ('Downloaded: ' + zip_file +"!")
 
print ("Extracting: " + zip_file + " ...")
zip_ref = zipfile.ZipFile(zip_file, 'r')
zip_ref.extractall(zip_dir)
zip_ref.close()
print ("Extraction Finished!")
 
print("#####")
```
