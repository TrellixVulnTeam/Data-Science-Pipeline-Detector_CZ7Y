```python
# Investigate age column - looks like year could have been inputted instead of age
pd.set_option('display.max_rows', None) #make sure all rows are outputted
train.groupby(['age']).count().sort_values(by = 'age', ascending = False)
```

```python
# Fix age data that looks like year. Credit to kanoki.org - https://kanoki.org/2019/07/17/pandas-how-to-replace-values-based-on-conditions/
train.loc[(train.age >= 1000),'age'] = 2014 - train.age
train.groupby(['age']).count().sort_values(by = 'age', ascending = False)
```
