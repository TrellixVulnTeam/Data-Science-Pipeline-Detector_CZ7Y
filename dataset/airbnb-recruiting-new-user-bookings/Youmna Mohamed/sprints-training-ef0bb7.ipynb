# **3. Data Pre-Processing**

## **3.1. Ahmed --->>> gender, sign-up method & language**

# **gender**

```python
# Explore gender
td.gender
```

```python
# Code the nominal values
td.gender.replace("MALE", 1, inplace = True)
td.gender.replace("FEMALE", 0, inplace = True)
td.gender.replace('-unknown-', np.nan, inplace=True)
td.gender.isnull().sum()
```

# **signup_method**

```python
# Explore signup_method
td.signup_method.describe
td.signup_method.isnull().sum()
td.signup_method.unique()
```

```python
# Code the nominal values
td.signup_method.replace("basic", 0, inplace = True)
td.signup_method.replace("facebook", 1, inplace = True)
td.signup_method.replace("google", 2, inplace = True)
```

# **language**

```python
# Explore language
td.language.isnull().sum()
td.language.describe()
# So we have 25 languages
td.language.unique()
```

```python
# Code nominal values
# Create a list with coded languages
languages_no = td.language.unique()
for i in range(25):
    languages_no[i] = i

# Replace
td.language.replace(to_replace = td.language.unique(), value = languages_no, inplace = True)
td.language.unique()
```

# **Age**

```python
td.age.describe()
```

```python
td['age'].fillna(td['age'].mean(),inplace =True)
Q1 = td['age'].quantile(0.25)
Q3 = td['age'].quantile(0.75)
IQR = Q3-Q1
l = Q1 - 1.5*IQR
u = Q3 + 1.5*IQR
a = td['age']< l
b = td['age']> u
td = td [~(a|b)]
td.describe()
```
