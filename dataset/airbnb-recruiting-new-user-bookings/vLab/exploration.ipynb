```python
train_users = pd.read_csv('../input/train_users_2.csv')
train_users.gender.replace('-unknown-', np.nan, inplace=True)
```

Pourcentage de valeurs manquantes de l'ensemble d'apprentissage, par colonne:

```python
users_nan = (train_users.isnull().sum() / train_users.shape[0]) * 100 
users_nan[users_nan > 0]
```

On voit que 4 colonnes seulement sont concernées par les données manquantes. 
Comparons la proportion de valeurs manquantes de "date_first_booking" avec la proportion de "NDF" sur la colonne "country_destination"

```python
train_users[train_users.country_destination=='NDF'].count()
```

```python
train_users[(train_users.date_first_booking.isnull()) & (train_users.country_destination != 'NDF')].count()
```

On voit ici que les valeurs manquantes de la colonne "date_first_booking" correspondent à la valeur de destination "NDF".

Pourcentage de valeurs manquantes de l'ensemble de test, par colonne:

```python
test_users = pd.read_csv('../input/test_users.csv')
test_users.gender.replace('-unknown-', np.nan, inplace=True)
```

```python
test_users_nan = (test_users.isnull().sum() / test_users.shape[0]) * 100 
test_users_nan[users_nan > 0]
```

On voit sans surprise que la colonne "date_first_booking" est vide. Les trois autres colonnes concernées présentent des valeurs manquantes dans des proportions plus ou moins similaires à celles de l'ensemble d'apprentissage.
