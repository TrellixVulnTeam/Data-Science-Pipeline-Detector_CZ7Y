```python
train_data = pd.read_csv('../input/airbnb-recruiting-new-user-bookings/train_users_2.csv.zip')
train_data.head()
```

```python
train_data.info()
```

```python
train_data.age.describe()
```

```python
print(sum(train_data.age > 100))
print(sum(train_data.age < 100))
```

```python
train_data[train_data.age > 120]['age'].describe()
```

```python
train_data['age'] = train_data['age'].apply(lambda x: np.nan if x > 100 else x)
print(sum(train_data.age > 100))
print(sum(train_data.age < 100))
```

```python
fig, ax = plt.subplots()
destination_percentage = train_data.country_destination.value_counts() / train_data.shape[0] * 100
destination_percentage.plot(kind='bar',color='#3498DB')
plt.xlabel('Destination Country')
plt.ylabel('Percentage')
sns.despine()
```

```python
fig, ax = plt.subplots()
pd.to_datetime(train_data['date_account_created']).value_counts().plot(kind='line', linewidth=1.2)
plt.xlabel('Date account created - line plot ')
sns.despine()
```

```python
train_data['timestamp_first_active'] = pd.to_datetime(train_data['timestamp_first_active'])
train_data['timestamp_first_active'] = train_data.timestamp_first_active.dt.weekday
train_data.timestamp_first_active.describe()
```

```python
train_data[train_data.timestamp_first_active == 3]['timestamp_first_active'].describe()
```

```python
fig, ax = plt.subplots()
signup_flow_percentage = train_data.signup_flow.value_counts() / train_data.shape[0] * 100
signup_flow_percentage.plot(kind='bar',color='#3498DB')
plt.xlabel('signup_flow')
plt.ylabel('Percentage')
sns.despine()
```
