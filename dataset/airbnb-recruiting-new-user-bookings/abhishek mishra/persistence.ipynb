# What does each columns tell
1. signup_method = How the users are signing at AirBnB. It consists of three unique categories ('facebook', 'google', and 'basic email signup')
2. signup_flow = This field is made from [ 0,  1,  2,  3,  4,  5,  6,  8, 10, 12, 15, 16, 20, 21, 23, 24, 25]. It reflects the page that a user came from for signup.
3. affiliate_channel: Which kind of paid advertisment attracted the user to come to the site (my understanding). It consists of following elements: ['api', 'content', 'direct', 'other', 'remarketing', 'sem-brand', 'sem-non-brand', 'seo']
4. affiliate_provider: Who was give this paid advertisement. It consists of these elements: ['baidu', 'bing', 'craigslist', 'daum', 'direct', 'email-marketing', 'facebook', 'facebook-open-graph', 'google', 'gsp', 'meetup', 'naver', 'other', 'padmapper', 'vast', 'wayn', 'yahoo', 'yandex'] 
5. first_affiliate_tracked: I don't have any idea what is it. It consists of following unique elements ['linked', 'local ops', 'marketing', 'omg', 'product', 'tracked-other', 'untracked']
6. signup app: How did the users signed up with AirBnB. It consists of following four elements: ['Android', 'Moweb', 'Web', 'iOS']
7. first_device_type: What device the users try to connect to AirBnb. It consists of following field: ['Android Phone', 'Android Tablet', 'Desktop (Other)', 'Mac Desktop', 'Other/Unknown', 'SmartPhone (Other)', 'Windows Desktop', 'iPad', 'iPhone']
8. First Browser: What browser that the users first use to connect to AirBnb. 
9. Country Destination: This we need to predict: The destination countries of users.
