[The COVID Tracking Project](https://covidtracking.com/) collects information from 50 US states, the District of Columbia, and 5 other US territories to provide the most comprehensive testing data we can collect for the novel coronavirus, SARS-CoV-2. We attempt to include positive and negative results, pending tests, and total people tested for each state or district currently reporting that data.

You can dowload data here: https://covidtracking.com/data/
