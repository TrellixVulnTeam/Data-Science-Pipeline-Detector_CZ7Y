```python
data = pd.read_csv("../input/coleridgeinitiative-show-us-the-data/train.csv")
data
```

```python
plt.figure(figsize = (15,50)),
sns.set_theme()
sns.countplot(y=data["cleaned_label"], order = data["cleaned_label"].value_counts().index, palette = "winter")
plt.ylabel("CLEANSED LABEL",fontsize = 50)
plt.show()
```

```python
from nltk.corpus import stopwords
stoplist = stopwords.words('english') + ['though']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
c_vec = CountVectorizer(stop_words=stoplist, ngram_range=(2,2))
# matrix of ngrams
ngrams = c_vec.fit_transform(data['cleaned_label'])
# count frequency of ngrams
count_values = ngrams.toarray().sum(axis=0)
# list of ngrams
vocab = c_vec.vocabulary_
df_bigram = pd.DataFrame(sorted([(count_values[i],k) for k,i in vocab.items()], reverse=True)
            ).rename(columns={0: 'frequency', 1:'bigram'})
```

```python
plt.figure(figsize = (15,80))
sns.lineplot(y = df_bigram["bigram"], x = df_bigram["frequency"])
plt.xticks(rotation = 90)
plt.ylabel("BIGRAM",fontsize = 50)
plt.show()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
c_vec = CountVectorizer(stop_words=stoplist, ngram_range=(3,3))
# matrix of ngrams
ngrams = c_vec.fit_transform(data['cleaned_label'])
# count frequency of ngrams
count_values = ngrams.toarray().sum(axis=0)
# list of ngrams
vocab = c_vec.vocabulary_
df_trigram = pd.DataFrame(sorted([(count_values[i],k) for k,i in vocab.items()], reverse=True)
            ).rename(columns={0: 'frequency', 1:'trigram'})
```
