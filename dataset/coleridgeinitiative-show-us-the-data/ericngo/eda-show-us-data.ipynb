## Word Cloud of label

````python
from wordcloud import WordCloud, STOPWORDS

words_in_titles = list(data.pub_title.str.split(expand=True).stack())

wordcloud = WordCloud(stopwords = STOPWORDS,
                      background_color = "white",
                      width = 3000,
                      height = 2000
                     ).generate(' '.join(words_in_titles))
plt.figure(1, figsize = (18, 12))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
````

## Document analysis

````python
import json

# read text
text = []
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        if filename.endswith('.json'):
            text.append(json.load(open(os.path.join(dirname, filename))))
````

````python
# section length
from collections import Counter
section_length = Counter([len(x) for x in text])
sections = sorted(section_length.keys())
plt.figure(figsize = (15,15))
plt.bar(sections, [section_length[x] for x in sections])
plt.show()
````
