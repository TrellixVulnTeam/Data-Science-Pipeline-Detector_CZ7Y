```python
#pub_title vs Ids
pub_titl_ids = {}
for titl in list(train['pub_title'].unique()):
    if train[train['pub_title']==titl]['Id'].nunique() > 1:
        pub_titl_ids[titl]= list(set(train[train['pub_title']==titl]['Id']))

pub_titl_ids
```

```python
#dataset_title vs dataset_label
titl_lab = {}
for titl in list(train['dataset_title'].unique()):
    if train[train['dataset_title']==titl]['dataset_label'].nunique() > 1:
        titl_lab[titl]= list(set(train[train['dataset_title']==titl]['dataset_label']))

titl_lab
```
