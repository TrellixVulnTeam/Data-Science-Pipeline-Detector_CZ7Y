```python
titles_list = list(df_train['dataset_label'])
commontitles =FreqDist(titles_list).most_common(25)
print(commontitles)

x, y = zip(*commontitles)
plt.figure(figsize=(60,40))
#plt.margins(0.02)
plt.bar(x, y)
plt.xlabel('Dataset titles', fontsize=50)
plt.ylabel('Frequency of Dataset Labels', fontsize=50)
plt.yticks(fontsize=40)
plt.xticks(rotation=60, fontsize=40, ha='right')
plt.title('Freq of 25 Most Common Dataset labels', fontsize=50)
plt.show()
```

```python
titles_list = list(df_train['dataset_title'])
commontitles =FreqDist(titles_list).most_common(45)
print(commontitles)

x, y = zip(*commontitles)
plt.figure(figsize=(70,40))
#plt.margins(0.02)
plt.bar(x, y)
plt.xlabel('Dataset titles', fontsize=50)
plt.ylabel('Frequency of Dataset Titles', fontsize=50)
plt.yticks(fontsize=40)
plt.xticks(rotation=60, fontsize=40, ha='right')
plt.title('Freq of 25 Most Common Dataset titles', fontsize=50)
plt.savefig("test.png")
plt.show()
```
