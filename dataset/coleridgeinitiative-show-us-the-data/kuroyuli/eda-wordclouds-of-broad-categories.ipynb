# <span style="color: green">WordClouds of broad categories</span>

```python
from collections import defaultdict

# adding extra stopwords to make more meaningful wordclouds
extra_stopwords = {'Education', 'Educational', 'Student', 'Students', 'Teacher', 'Teachers', 'School', 'Schools', 'Study', 'Studys', 'Science', 'Ocean', 'Aging', 'Age', 'Alzheimer', 'Alzheimer s', 'Alzheimer Disease', 'Agriculture', 'Agricultural', 'Research', 'Development', 'Mathematic', 'Mathematics', 'Rural', 'Urban', 'bird', 'birds', 'COVID', 'CoV', 'using', 'Analysis', 'effect', 'effects', 'Survey', 'among', 'data', 'D', 'S', 'U'}
STOPWORDS = STOPWORDS | extra_stopwords

words_in_category = defaultdict(list)

for _, row in train.iterrows():
    words_in_category[row['category']].extend(row['pub_title'].split())

def wordcloud_draw(data, color = 'black', colormap="summer"):
    wordcloud = WordCloud(stopwords = STOPWORDS,
                          background_color = color,
                          colormap=colormap,
                          width = 1000,
                          height = 500
                         ).generate(' '.join(data))
    plt.figure(1, figsize = (40, 30))
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.show()

for category in train['category'].unique():
    print("Wordcloud for", category, ":")
    wordcloud_draw(words_in_category[category])
```
