```python
import pandas as pd
import glob
import json
import re
import plotly.express as exp
import os
import numpy as np
from tqdm.autonotebook import tqdm

def jaccard(str1, str2): 
    a = set(str1.lower().split()) 
    b = set(str2.lower().split())
    c = a.intersection(b)
    return float(len(c)) / (len(a) + len(b) - len(c))
```

```python
def clean_text(txt):
    return re.sub('[^A-Za-z0-9]+', ' ', str(txt).lower()).strip()
```

```python
def find_dataset(text):
    f = []
    for x in train_datasets:
        if text.find(x)!=-1:
            f.append(x+"|")
    return "".join(f)[:-1]     
```

```python
submission = pd.read_csv("../input/coleridgeinitiative-show-us-the-data/sample_submission.csv")
```

```python
test_path = "../input/coleridgeinitiative-show-us-the-data/test"
```

```python
for i in tqdm(range(len(submission))):
    text = json.load(open(os.path.join(test_path,submission.iloc[i,0]+".json")))
    sec = []
    for x in text:
        sec.append(x['section_title'])
        sec.append(" ")    
        sec.append(x['text'])    
    entire = "".join(sec)
    entire = clean_text(entire)
    submission.loc[i,"PredictionString"] = find_dataset(entire)
```

```python
submission.to_csv("submission.csv",index=False)
```
