```python
import glob
import numpy as np
import pandas as pd
from termcolor import colored, cprint

train = pd.read_csv('../input/coleridgeinitiative-show-us-the-data/train.csv')
```

```python
train
```

```python
#Path to the JSON files
train_files = glob.glob("../input/coleridgeinitiative-show-us-the-data/train/*.json")
test_files = glob.glob("../input/coleridgeinitiative-show-us-the-data/test/*.json")
```

```python
#using random index for demo
idx = 1

ID = train_files[idx].split('/')[-1][:-5]
df_id = train[train['Id']==ID]
df_id
```

```python
dataset_title = df_id['dataset_title']
dataset_title = list(set(dataset_title.tolist()))
dataset_title
```

```python
json_content = pd.read_json(train_files[idx])
json_content
```

```python
section_titles = list(json_content['section_title'].to_numpy())
section_titles
```

```python
string = ''

#appending the text from all sections 
for t in section_titles:
    string += json_content[json_content['section_title'] == t]['text'].item()
    string += '\n'
    
    
start_idx = string.find(dataset_title[0])


print(string[:start_idx]  +  
      colored(string[start_idx:start_idx+len(dataset_title[0])], 'red', attrs=['reverse', 'blink']) +   #highlighting the dataset name in the text
      string[start_idx+len(dataset_title[0]):]
     )
```
