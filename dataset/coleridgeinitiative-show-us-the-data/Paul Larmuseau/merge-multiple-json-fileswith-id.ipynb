# Loading all the JSON files

 #articles are splitted in paragraphs
 

```python
# load json file using pandas
def readjson(data_files,variab):
    df_train = pd.DataFrame()
    for count,ele in enumerate(data_files,len(data_files)):
        #df_train = pd.concat([df_train, pd.read_json(ele)])
        pdele=pd.read_json(ele).reset_index()
        idnr=ele.split('/')[-1:][0].split('.')[0]
        pdele['Id']=idnr
        df_train=df_train.append(pdele)
    df_train.to_csv(variab+'.csv',index=False)
    print(df_train.head())
    return df_train

testtxt=readjson(test_files,'test')
traintxt=readjson(train_files,'train')
testtxt
```

```python
traintxt
```

```python
submit=testtxt.merge(train,how='inner',left_on='Id',right_on='Id')
submit
```

```python
submit=submit.groupby('Id').max().reset_index()[['Id','cleaned_label']]
submit.columns=['Id','PredictionString']
submit.to_csv('submission.csv', index = False)
```
