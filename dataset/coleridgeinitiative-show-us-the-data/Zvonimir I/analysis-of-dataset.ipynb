## Most common noins and verbs in titles

```python
nlp = spacy.load("en_core_web_sm")
```

```python
noun_counter = Counter()
verb_counter = Counter()
```

```python
tokenized = train.pub_title.apply(nlp)
```

```python
for t in tokenized:
    for token in t:
        if token.pos_ == 'NOUN':
            noun_counter.update([token.text.lower()])
        if token.pos_ == 'VERB':
            verb_counter.update([token.text.lower()])
```

```python
noun_counter.most_common()
```

```python
verb_counter.most_common()
```
