```python
import urllib.request, json 
import pandas as pd
#from pandas.io.json import json_normalize
############## https://labs.data.gov/dashboard/offices/qa
api_list = ["http://www.usda.gov/data.json",
            "http://www.commerce.gov/data.json",
            #"https://www.defense.gov/data.json",
            "http://www.ed.gov/data.json",
            "http://www.energy.gov/data.json",
            "http://www.hhs.gov/data.json",
            "http://www.dhs.gov/data.json",
            "http://www.hud.gov/data.json",
            "http://www.usdoj.gov/data.json",
            "http://www.dol.gov/data.json",
            #"http://www.state.gov/data.json",
            "http://www.ustreas.gov/data.json",
            "http://www.dot.gov/data.json",
            "http://www.va.gov/data.json"
            "http://www.gsa.gov/data.json",
            "http://www.nsf.gov/data.json",
            #"http://www.nrc.gov/data.json",
            "http://www.opm.gov/data.json",
            "http://www.sba.gov/data.json",
            "http://www.ssa.gov/data.json",
            "http://www.usaid.gov/data.json",
            "http://www.usccr.gov/data.json",
            "http://www.consumerfinance.gov/data.json",
            "http://www.cpsc.gov/data.json",
            "http://www.eeoc.gov/data.json",
            "http://www.fdic.gov/data.json",
            #"http://www.fhfa.gov/data.json",
            "http://www.frtib.gov/data.json",
            "http://www.ftc.gov/data.json",
            #"http://www.mspb.gov/data.json",
            "https://www.doi.gov/sites/doi.gov/files/data.json",
            "http://www.udall.gov/data.json",
            "http://www.ncua.gov/data.json",
            #"http://www.arts.gov/data.json",
            "http://www.neh.gov/data.json",
            "http://www.ntsb.gov/data.json",
            "http://www.osc.gov/data.json",
            "http://www.pbgc.gov/data.json",
            "http://www.presidio.gov/data.json",
            "https://www.pdsdc.org/data.json",
            "http://www.sec.gov/data.json",
            "http://www.usitc.gov/data.json", 
           ]
#api_list = ["http://www.commerce.gov/data.json"]
i=0
for site in api_list:
    print (site)
    with urllib.request.urlopen(site) as url:
        try:
            data = json.loads(url.read().decode('utf8'))
        except json.decoder.JSONDecodeError:
            pass
        
        #print (data)
        if i==0:
            df = pd.DataFrame(pd.json_normalize(data, 'dataset'))
        else:
            temp = pd.DataFrame(pd.json_normalize(data, 'dataset'))
            df = df.append(temp, ignore_index=True)
    i=i+1
#df=df[df['@type'].str.contains('dcat:Dataset')]
df
```

```python
df2 = df[['title']].copy()
df2['title'].str.replace(",","")
df2['title']=df['title'].str.lower()
df["title"] = df['title'].str.replace('[^\w\s]','')

df
```
