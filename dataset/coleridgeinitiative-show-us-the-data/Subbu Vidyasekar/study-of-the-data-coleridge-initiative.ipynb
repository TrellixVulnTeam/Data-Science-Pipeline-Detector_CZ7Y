```python
import plotly.express as px
import json
```

```python
train = pd.read_csv("/kaggle/input/coleridgeinitiative-show-us-the-data/train.csv")
train.head()
```

```python
train.shape
```

```python
train.dataset_title.value_counts()
title_count = train.groupby('dataset_title')['Id'].count().reset_index(name = 'counts')
title_count = title_count.sort_values(by=['counts'], ascending = False)
fig = px.bar(title_count[0:10], x = 'dataset_title', y = 'counts')
fig.show()
```

```python
title_count = train.groupby('dataset_label')['Id'].count().reset_index(name = 'counts')
title_count = title_count.sort_values(by=['counts'], ascending = False)
fig = px.bar(title_count[0:10], x = 'dataset_label', y = 'counts')
fig.show()
```
