```python
sample_subm = pd.read_csv('../input/m5-forecasting-accuracy/sample_submission.csv')
```

```python
sample_subm = pd.DataFrame(sample_subm.id)
sample_subm.col = 'id'
```

```python
sample_subm.tail()
```

```python
evalSubmission = pd.read_csv('../input/m5-lgbm-testing/evalSubmission.csv')
valSubmission = pd.read_csv('../input/m5-lgbm-testing/valSubmission.csv')
```

```python
submColumns = [f"F{val}" for val in range(1, 29)]
submColumns.insert(0, 'id')
```

```python
evalSubmission.columns = submColumns
valSubmission.columns = submColumns
```

```python
evalSubmission.head()
```

```python
submission = pd.concat([valSubmission, evalSubmission], axis=0) 
```

```python
final_subm = sample_subm.merge(submission, how='left', on='id')
```

```python
final_subm.to_csv('final_subm.csv', index=False)
```
