***In this competition, there are many zero sells in item level***.

Let's explore it!

```python
import os
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../input/m5-forecasting-accuracy/sales_train_evaluation.csv")
dg = df.iloc[:, -28:]
```

```python
dg.head()
```

## How many zeros in each item of all data?

```python
a = np.array((df == 0).astype(int).sum(axis=1))
df['zeros'] = a
df['percent'] = df['zeros']/1941
```

```python
df['percent'].plot.hist()
```

## How many zeros in each item of evaluation data?

```python
a = np.array((dg == 0).astype(int).sum(axis=1))
dg['zeros'] = a
dg['percent'] = dg['zeros']/28
```

```python
dg.head()
```

```python
dg['percent'].plot.hist()
```
