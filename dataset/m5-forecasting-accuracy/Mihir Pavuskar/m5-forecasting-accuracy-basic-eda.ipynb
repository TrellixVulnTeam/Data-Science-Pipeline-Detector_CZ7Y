## 2.1 Sales Data
* This dataset represents the number of a given item bought from a given store at a given day
* The attributes being
    * item_id: The id of the product.
    * dept_id: The id of the department the product belongs to.
    * cat_id: The id of the category the product belongs to.
    * store_id: The id of the store where the product is sold.
    * state_id: The State where the store is located.
    * d_1, d_2, …, d_i, … d_1941: The number of units sold at day i, starting from 2011-01-29.


```python
sales = pd.read_csv('../input/m5-forecasting-accuracy/sales_train_validation.csv')
sales.head()
```

```python
sales.shape
```

* 30,490 is a lot of rows. Especially for plotting and stuff atleast as far as EDA is concerned 

```python
sales.describe()
```

### 2.1.1 days d_1 to d_1913
* What we can infer from this is that the data is highly skewed.
* The mean of almost all day is close to 1, min being 0 but the max being 360
* This means most items were hardly bought on a given day, except for certain items which were bought in bulk
* To visualize this better, how about a plot

```python
sales['d_1'].iplot()
```

```python
sales['d_1'].value_counts().iplot()
```

* The the count plot above makes it even more clear that the data is highly skewed to the left with the highest frequency beimg 0.

### 2.1.2 dept_id and cat_id
* The visualizations below show the distribution of items in a given department
* Food catagory has the most departments and thus the most items, followed by household and then by hobbies
