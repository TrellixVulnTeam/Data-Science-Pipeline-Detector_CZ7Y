```python
data["F1"] = data["F1"] - data["F1"]*0.016
data["F2"] = data["F2"] + data["F2"]*0.017
data["F3"] = data["F3"] - data["F3"]*0.018
data["F4"] = data["F4"] + data["F4"]*0.019
data["F5"] = data["F5"] - data["F5"]*0.020
data["F6"] = data["F6"] + data["F6"]*0.021
data["F7"] = data["F7"] - data["F7"]*0.022
data["F8"] = data["F8"] + data["F8"]*0.023
data["F9"] = data["F9"] - data["F9"]*0.024
data["F10"] = data["F10"] + data["F10"]*0.025
 
```
