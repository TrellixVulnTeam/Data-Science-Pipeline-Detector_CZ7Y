# DATA VISUALIZATION

```python
temp = train_x['Units'].head(100)
sns.boxplot(temp).set_title("DISTRIBUTION OF UNITS SOLD")
```

```python
plt.bar(x=train_x['cat_id'],height=train_x['Units'])
plt.xlabel("CATEGORIES")
plt.ylabel("UNITS SOLD")
plt.title("UNITS SOLD CATEGORY WISE")
plt.show()
```

**This Plot clearly shows that most number of good solds is from the Food Category**

- Let's try something for state's now. 

```python
plt.bar(x=train_x['state_id'],height=train_x['Units'])
plt.xlabel("STATES")
plt.ylabel("UNITS SOLD")
plt.title("UNITS SOLD STATES WISE")
plt.show()
```

- CA stands for California

- WI Stands for Wisconsin

- TX Stands for Texas.

- UNITS SOLD ORDER: [ CA(CALIFORNIA) > TX(TEXAS) > WI(WISCONSIN) ]

```python
sns.barplot(x='cat_id',y='Units',
           hue='state_id',
           data=train_x).set_title("UNITS SOLD IN STATES IN DIFFERENT CATEGORIES")
```

## INSIGHTS:

- Clearly in all three states Category **Food is most Sold**.

- CA(California) has the Most units sold in all three categories.
