# give different score with different weights

```python
poisson = poisson.sort_values(by = 'id').reset_index(drop = True)
tweedie = tweedie.sort_values(by = 'id').reset_index(drop = True)
Rv2 = Rv2.sort_values(by = 'id').reset_index(drop = True)
best_score = best_score.sort_values(by = 'id').reset_index(drop = True)
lig = lig.sort_values(by = 'id').reset_index(drop = True)
sub = poisson.copy()

for i in sub.columns :
    if i != 'id' :
        sub[i] = 0.5 * best_score[i] + 0.3 * tweedie[i] + 0.2 * poisson[i]
        
sub.to_csv('submission.csv', index = False)
sub.head()
```
