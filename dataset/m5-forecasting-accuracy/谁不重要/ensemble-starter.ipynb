# Data Import

```python
df1 = pd.read_csv("../input/m5-forecasting-ensemble-data/submission_084918.csv")
df2 = pd.read_csv("../input/m5-forecasting-ensemble-data/submission_085051.csv")
df3 = pd.read_csv("../input/m5-forecasting-ensemble-data/submission_084855.csv")
```

# The Ensemble

```python
id_col = df1[['id']]
id_col.head()
```

```python
df1.drop(['id'], axis=1, inplace=True)
df2.drop(['id'], axis=1, inplace=True)
df3.drop(['id'], axis=1, inplace=True)
```

```python
df1.head() # example of output
```

```python
# number of datasets for ensemble
N = 3 

# either divide ensemble by number of components or assign them weights that sum to 1
w1 = 0.33
w2 = 0.33
w3 = 0.34
```

```python
ensemble = w1*df1 + w2*df2 + w3*df3
```

```python
predictions = pd.concat([id_col, ensemble], axis=1)
```

# Make Submission
