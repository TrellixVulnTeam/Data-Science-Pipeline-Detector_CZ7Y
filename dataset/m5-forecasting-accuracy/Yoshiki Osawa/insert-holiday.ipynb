```python
df_calendar = pd.read_csv("../input/m5-forecasting-accuracy/calendar.csv")
```

```python
df_calendar["holiday_CA"] = 0
df_calendar["holiday_TX"] = 0
df_calendar["holiday_WI"] = 0
```

```python
us_holidays = holidays.UnitedStates()

print()
print("----- CA -----")
# # Print all the holidays in CA state of UnitedStates in year 2018 
for ptr in holidays.UnitedStates(state='CA', years = 2018).items(): 
    print(ptr)
    
print()
print("----- TX -----")
# Print all the holidays in TX state of UnitedStates in year 2018 
for ptr in holidays.UnitedStates(state='TX', years = 2018).items(): 
    print(ptr)

print()
print("----- WI -----")
# Print all the holidays in WI state of UnitedStates in year 2018 
for ptr in holidays.UnitedStates(state='WI', years = 2018).items(): 
    print(ptr)
```

```python
for year in range(2011, 2017):
    for date in sorted(holidays.US(state='CA', years=year)):
        holiday = date
        str_holiday = holiday.strftime('%Y-%m-%d')
        df_calendar.loc[df_calendar['date'] == str_holiday, 'holiday_CA'] = 1
        
for year in range(2011, 2017):
    for date in sorted(holidays.US(state='TX', years=year)):
        holiday = date
        str_holiday = holiday.strftime('%Y-%m-%d')
        df_calendar.loc[df_calendar['date'] == str_holiday, 'holiday_TX'] = 1

for year in range(2011, 2017):
    for date in sorted(holidays.US(state='WI', years=year)):
        holiday = date
        str_holiday = holiday.strftime('%Y-%m-%d')
        df_calendar.loc[df_calendar['date'] == str_holiday, 'holiday_WI'] = 1
```
