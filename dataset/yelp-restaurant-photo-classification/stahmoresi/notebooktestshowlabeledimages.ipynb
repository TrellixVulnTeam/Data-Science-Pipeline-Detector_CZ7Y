```python
train_attr = pd.read_csv('../input/train.csv')
#train_attr['labels_list'] = train_attr['labels'].str.split(' ')
train_attr
```

```python
# find all the restaurants that are good for lunch (label=0)
train_attr['is_goodforlunch'] = train_attr['labels'].str.contains('4')
goodforlunch_businesses = train_attr[train_attr.is_goodforlunch==True].business_id.tolist()
len(goodforlunch_businesses)
```

```python
goodforlunch_businesses[0]
```

```python
goodforlunch_photos = train_photo_to_biz_ids[train_photo_to_biz_ids.business_id.isin(goodforlunch_businesses)].photo_id.tolist()
len(goodforlunch_photos)
```

```python
#goodforlunch_photos1 = train_photo_to_biz_ids[train_photo_to_biz_ids.business_id==1001].photo_id.tolist()
goodforlunch_photos1 = train_photo_to_biz_ids[train_photo_to_biz_ids.business_id==goodforlunch_businesses[0]].photo_id.tolist()
len(goodforlunch_photos1)
```

```python
import math

num_images_for_show_column = 4
rows= math.ceil(len(goodforlunch_photos1)/num_images_for_show_column)
#photos_to_show = np.random.choice(goodforlunch_photos1,num_images_for_show**2)

for x in range(len(goodforlunch_photos1)):
#for x in range(num_images_for_show ** 2):
#for x in len(goodforlunch_photos1):
        
        plt.subplot(rows,num_images_for_show_column , x+1)
        im = Image.open(os.path.join('../input/','train_photos',''.join([str(goodforlunch_photos1[x]),'.jpg'])))
        plt.imshow(im)
        plt.axis('off')
```
