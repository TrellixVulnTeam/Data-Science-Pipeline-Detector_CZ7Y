```python
covid_data_train.dropna(axis=1, inplace= True)
covid_data_test.dropna(axis=1, inplace= True)
```

```python
import plotly.express as px
```

```python
import plotly.express as px
fig= plt.figure(figsize=(45,30))
fig= px.pie(covid_data_train, names= 'Country_Region', values= 'TargetValue', color_discrete_sequence= px.colors.sequential.RdBu, hole= 0.4)
fig.update_traces(textposition = 'inside')

fig.show()
```

```python
import seaborn as sns
sns.barplot(y= 'TargetValue', x='Target', data= covid_data_train)
plt.show()
```

```python
sns.barplot(y= 'Population', x='Target', data= covid_data_train)
plt.show()
```

```python
grouped_data= covid_data_train.groupby('Country_Region').sum()

top_10_pop_countries= grouped_data.nlargest(10,'Population')['TargetValue']

top_10_pop_countries
```

```python
fig= px.bar(x= top_10_pop_countries.index, y= top_10_pop_countries.values, title= 'top 10 populous countries vs number of covid cases', labels= dict(x= 'Countries', y='number of covid cases'))
fig.show()
```
