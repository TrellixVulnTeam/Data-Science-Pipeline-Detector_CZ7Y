# Aerial Cactus Identification Submission



# Import Statements

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import zipfile
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

# Load DataFrame

```python
test = pd.read_csv('../input/aerial-cactus-identification/sample_submission.csv')

print('Test Set Size:', test.shape)
```

```python
test.head()
```

# Extract Test Images

```python
zip_ref = zipfile.ZipFile('/kaggle/input/aerial-cactus-identification/test.zip')
zip_ref.extractall()
```

```python
test_path = "test/"
print('Test Images:', len(os.listdir(test_path)))
```

# Data Generator

```python
BATCH_SIZE = 64

test_datagen = ImageDataGenerator(rescale=1/255)

test_loader = test_datagen.flow_from_dataframe(
    dataframe = test,
    directory = test_path,
    x_col = 'id',
    batch_size = BATCH_SIZE,
    shuffle = False,
    class_mode = None,
    target_size = (32,32)
)
```

# Load Model

```python
cnn = keras.models.load_model('../input/models/Cactus/cactus_model_v01.h5')
cnn.summary()
```

# Test Predictions

```python
test_probs = cnn.predict(test_loader)
print(test_probs.shape)
```

```python
print(test_probs[:10,].round(2))
```

# Prepare Submission

```python
submission = pd.read_csv('../input/aerial-cactus-identification/sample_submission.csv')
submission.head()
```

```python
submission.has_cactus = test_probs[:,1]
submission.head()
```

```python
submission.to_csv('submission.csv', header=True, index=False)
```
