**Load Train Data and create a train test split/ 80%train 20%test**

```python
train_df = pd.read_csv("../input/train.csv")
train_df.head()

x_train,x_test,y_train,y_test = train_test_split(
    train_df['id'],
    train_df['has_cactus'],
    test_size = 0.2,
    random_state = 1)
```

**Load in Train and test Images**

```python
X_train = []
for images in tqdm(x_train):
    img = plt.imread('../input/train/train/' + images)
    X_train.append(img)
    

X_test = []
for images in tqdm(x_test):
    img = plt.imread('../input/train/train/' + images)
    X_test.append(img)
    
X_train = np.array(X_train)
X_test = np.array(X_test)

X_train = X_train.astype('float32')
X_test = X_test.astype('float32')

X_train = X_train/255
X_test = X_test/255
```

**Data Augmentation**

```python
augmentations = ImageDataGenerator(
    vertical_flip=True,
    horizontal_flip=True,
    zoom_range=0.1)

augmentations.fit(X_train)
```
