```python
from PIL import Image
from tqdm import tqdm
import os
```

```python
Images_DIR = "/kaggle/input/h-and-m-personalized-fashion-recommendations/images/"
Save_DIR = "images_resized/"

os.mkdir(Save_DIR)
directories = os.listdir(Images_DIR)
```

```python
for folder in tqdm(directories, position=0, leave=True):
    
    img_files = os.listdir(Images_DIR+folder+'/')
    
    for img_file in img_files:
         
        image = Image.open(Images_DIR+folder+'/'+img_file)
        image = image.resize((224, 224))
        
        if not os.path.isdir(Save_DIR+folder+'/'):
            os.mkdir(Save_DIR+folder+'/')
        
        image.save(Save_DIR+folder+'/'+img_file)
        
    print(f'Folder {folder} ... Next!')
```
