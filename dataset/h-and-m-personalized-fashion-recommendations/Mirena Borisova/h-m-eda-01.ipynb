```python
from os import walk

def show_articles(folder, no_images=3):
    folder_path = f'../input/h-and-m-personalized-fashion-recommendations/images/{folder}/'
    files = []
    
    for _, _, filenames in walk(folder_path):
        files.extend(filenames)
    no_files = len(files)
    if no_images > no_files:
        no_images = no_files
        print('WARNING! IN FOLDER IMAGES ARE LESS THAN REQUESTED')
        
    images = files[:no_images]
    fig, ax = plt.subplots(1, no_images, figsize=(12, 4))
    
    for i, image in enumerate(images):
        art_id = image.split('.')[0]
        image = plt.imread(folder_path+image)
        ax[i].imshow(image, aspect='equal')
        ax[i].grid(False)
        ax[i].set_xticks([], [])
        ax[i].set_yticks([], [])
        ax[i].set_xlabel(art[art['article_id'] == int(art_id[1:])]['prod_name'].iloc[0])
    
    plt.show()
    
show_articles('020', 4)
```

```python
import matplotlib.ticker as mtick

def plot_bar(database, col, figsize=(13, 5), pct=False, label='ARTICLES'):
    fig, ax = plt.subplots(figsize=figsize, facecolor='lemonchiffon')
    
    for spine in ['bottom', 'left']:
        ax.spines[spine].set_visible(True)
        ax.spines[spine].set_linewidth(2)
        ax.spines[spine].set_color('goldenrod')
        
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    
    if pct:
        data = database[col].value_counts()
        data = data.div(data.sum()).mul(100)
        data = data.reset_index()
        ax = sns.barplot(data=data, x=col, y='index', color='gold', lw=1.5, ec='olive', zorder=2)
        ax.set_xlabel(f'% OF {label}', color='goldenrod', fontsize=10, weight='bold')
        ax.xaxis.set_major_formatter(mtick.PercentFormatter())
    else:
        data = database[col].value_counts().reset_index()
        ax = sns.barplot(data=data, x=col, y='index', color='darkgoldenrod', lw=1.5, ec='olive', zorder=2)
        ax.set_xlabel(f'{label} NUMBER',color='gold', fontsize=10, weight='bold')
        
    ax.grid(zorder=0)
    ax.text(0, -0.75, col, color='goldenrod', fontsize=10, ha='left', va='bottom', weight='bold', style='italic')
    ax.set_ylabel('')
    
    plt.show()
```

```python
def show_items_in_category(column, value, no_images=4, title=None):
    data = art[art[column] == value]
    article_ids = data['article_id'].iloc[:no_images].to_list()
    
    fig, ax = plt.subplots(1, no_images, figsize=(12, 4))
    
    for i, article_id in enumerate(article_ids):
        folder = str(article_id)[:2]
        file_path = f'../input/h-and-m-personalized-fashion-recommendations/images/0{folder}/0{article_id}.jpg'
        
        image = plt.imread(file_path)
        ax[i].imshow(image, aspect='equal')
        ax[i].grid(False)
        ax[i].set_xticks([], [])
        ax[i].set_yticks([], [])
        ax[i].set_xlabel(art[art['article_id'] == int(article_id)]['prod_name'].iloc[0])
        
    fig.suptitle(title)
    plt.show()
```
