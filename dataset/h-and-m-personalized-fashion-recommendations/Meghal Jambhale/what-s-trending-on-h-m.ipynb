```python
import os
import numpy as np
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
```

<p style='font-family: Trebuchet MS'><font size=6>Let's explore the dataset. Following is the article dataset which consists of various columns. This is linked to the images by the 'article_id'. Hence we will find the corresponding images of those id's and add it to the article dataset.
</font>
</p>

```python
article=pd.read_csv("../input/h-and-m-personalized-fashion-recommendations/articles.csv")
print(article.columns)
```

```python
images_path=[]
i=0
for p in article['article_id'].tolist():
    path= '../input/h-and-m-personalized-fashion-recommendations/images/' + '0' + str(p)[:2] + '/'+ '0' + str(p) +'.jpg'

    if os.path.exists(path):
        i+=1
        images_path.append(path)
    else: images_path.append(None)
print(f'{i} articles have corresponding images')

article['images_path']=images_path
article = article[['article_id', 'prod_name', 'colour_group_name', 'images_path',
             'product_type_name', 'product_group_name', 'detail_desc']]
```

```python
article['prod_name'].describe()
```

<p style='font-family: Trebuchet MS'><font size=6>
    Now let's find out which colour&#127912 is trending in the H&M store!
</font>
</p>

```python
product_count = Counter(list(article['colour_group_name']))
prod_count=dict(product_count.most_common(10))
X=list(prod_count.keys())
y=prod_count.values()
colurs=['#000000','#003366','#ffffff','#ffe6e6','#b3b3b3','#f5f5dc','#00ccff','#e60000','#ccf5ff','#8A865D']
ax=plt.axes()
ax.set_facecolor('#ff6633')
plt.bar(X,y,color=colurs)
plt.xticks(X, X, rotation='vertical')
plt.title("TRENDING COLOURS")
plt.show()
```

<p style='font-family: Trebuchet MS'><font size=6>
    Now let's checkout which age group gets 'SHOPOHOLIC OF THE YEAR' award!
</font>
</p>

```python
customer=pd.read_csv("../input/h-and-m-personalized-fashion-recommendations/customers.csv")
customer.head()
```

```python
age_count = Counter(list(customer['age']))
age_count=dict(age_count.most_common(30))
age_count=pd.DataFrame({'Age': age_count.keys(),
                      'Count':age_count.values()})
age_count.head()
```

```python
# set figure size
plt.figure(figsize=(20,12))

# plot polar axis
ax = plt.subplot(111, polar=True)

# remove grid
plt.axis('off')

# Set the coordinates limits
upperLimit = 100
lowerLimit = 30

# Compute max and min in the dataset
max = age_count['Count'].max()

# Let's compute heights: they are a conversion of each item value in those new coordinates
# In our example, 0 in the dataset will be converted to the lowerLimit (10)
# The maximum will be converted to the upperLimit (100)
slope = (max - lowerLimit) / max
heights = slope * age_count.Count + lowerLimit

# Compute the width of each bar. In total we have 2*Pi = 360Â°
width = 2*np.pi / len(age_count.index)

# Compute the angle each bar is centered on:
indexes = list(range(1, len(age_count.index)+1))
angles = [element * width for element in indexes]
angles

# Draw bars
bars = ax.bar(
    x=angles, 
    height=heights, 
    width=width, 
    bottom=lowerLimit,
    linewidth=2, 
    edgecolor="white",
    color='#cc2900')

labelPadding=4

for bar, angle, height, label in zip(bars,angles, heights, age_count["Age"]):

    # Labels are rotated. Rotation must be specified in degrees :(
    rotation = np.rad2deg(angle)

    # Flip some labels upside down
    alignment = ""
    if angle >= np.pi/2 and angle < 3*np.pi/2:
        alignment = "right"
        rotation = rotation + 180
    else: 
        alignment = "left"

    # Finally add the labels
    ax.text(
        x=angle, 
        y=lowerLimit + bar.get_height() + labelPadding, 
        s=label, 
        ha=alignment, 
        va='center', 
        rotation=rotation, 
        rotation_mode="anchor")
```
