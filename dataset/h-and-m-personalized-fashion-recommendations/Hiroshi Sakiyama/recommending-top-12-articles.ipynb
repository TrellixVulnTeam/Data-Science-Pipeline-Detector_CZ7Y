# Displaying pictures

```python
dir_name = '../input/h-and-m-personalized-fashion-recommendations/images'

def show_jpg(file_name):
    folder_name = file_name
    file_path = dir_name + '/' + folder_name[:3]  + '/' + file_name + '.jpg'
    if file_name + '.jpg' in os.listdir(dir_name + '/' + folder_name[:3]):       
        img = cv2.imread(file_path)
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)) 
        plt.show()
    else:
        print('<<No Image>>')
```

```python
top12 = '0706016001 0706016002 0372860001 0610776002 0759871002 0464297007 0372860002 0610776001 0399223001 0706016003 0720125001 0156231001'
for i in range(12):
    file_name = top12[11 * i:11 * i + 10]
    folder_name = file_name
    print(file_name + '.jpg')
    show_jpg(file_name)
```
