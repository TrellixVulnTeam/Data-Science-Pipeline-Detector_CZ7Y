Let us look into all columns:

article id : uid for article

product_code and product_name for different products

product_type_name and product_type_no : they are parent group of product_name , ie every product_name belongs to some product_type_name

product_group_name : every product is part of some group

graphical appearence : for every cloth it has some graphical appearence and a unique code for each type of graphical appearence

color group code and color name : color of each product and each percieved color columns gives more info about color of the product

index code and index name : unique index and name for products

section name and section number : section which product belongs to

garment group name and number : type of garment with its unique id

desc : description for the product
