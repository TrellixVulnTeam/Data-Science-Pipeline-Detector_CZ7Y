```python
df = articles_per_customer.map(lambda x:"|".join(x)).str.get_dummies()
x = df.values
y = np.dot(x.T,x)
n_pairs = pd.DataFrame(y,index = df.columns,columns = df.columns)
np.fill_diagonal(n_pairs.values,0)
n_pairs
```

```python
df = transactions_merged.head(100000).groupby('postal_code').product_group_name.value_counts()
d = {}
max_val = {}
for x,y in df.index.values:
    if x not in d:
        d[x] = y
        max_val[x] = df[(x,y)]
    else:
        if df[(x,y)]>max_val[x]:
            max_val[x] = df[(x,y)]
            d[x] = y
top_group_per_code = pd.Series(d)
top_group_per_code.head(10)
```

```python
# sales distribution of top 100 selling articles
sns.set_style("darkgrid")
figure, ax = plt.subplots(figsize=(20, 5))
ax = sns.histplot(article_sorted_by_sales.values[:100], bins=100, color='xkcd:blue')
ax.set_xlabel('distribution of sales')
ax.set_ylabel('sales')
plt.show()
```

```python
# sales graph of articles sorted by sales volume
sns.set_style("darkgrid")
figure, ax = plt.subplots(figsize=(8,8))
ax = sns.lineplot(x=range(article_sorted_by_sales.shape[0]), y=article_sorted_by_sales.values, color='xkcd:blue')
ax.set_xlabel('article sales ranking')
ax.set_ylabel('sales')
plt.show()
```

```python
# how much top selling 1% articles contribute to overall sales
top_1p_articles = article_sorted_by_sales.head(int(article_sorted_by_sales.shape[0]*0.01))
top_1p_articles.sum()/ total_sales
```

```python
# how much top selling n% articles contribute to overall sales
n_pct_sales = [article_sorted_by_sales.head(int(article_sorted_by_sales.shape[0]*0.01*(n))).sum() / total_sales * 100 for n in range (1, 100+1)]
# n_pct_sales
```

```python
# top selling n% articles => contribution percentage to total sales
for n in range(100):
    print("{:3}% => {:3.2f} % sales (+{:3.2f})%".format(n+1, n_pct_sales[n], n_pct_sales[n] - n_pct_sales[n-1] if n>0 else 0))
    print("-------------------------------") if (n+1) % 10 == 0 and (n+1) != 0 else 0
```

```python
sns.set_style("darkgrid")
figure, ax = plt.subplots(figsize=(6,6))
ax = sns.lineplot(x=range(1, 100+1), y=n_pct_sales, color='xkcd:blue')
ax.set_xlabel('n percent articles')
ax.set_ylabel('sales')
plt.show()
```
