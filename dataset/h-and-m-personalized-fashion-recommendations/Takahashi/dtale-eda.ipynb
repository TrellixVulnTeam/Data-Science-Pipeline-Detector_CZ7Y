# Check image data

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

```python
def check_img(article_ids):
    len_articles = len(article_ids)-1
    f, ax = plt.subplots(len_articles//5+1, 5, figsize=(20,10))
    i = 0
    plt.subplots_adjust(wspace=0.4, hspace=0.6)
    for article_id in article_ids:
        desc = articles[articles['article_id'] == article_id]['detail_desc'].iloc[0]
        desc_list = desc.split(' ')
        for j, elem in enumerate(desc_list):
            if j > 0 and j % 5 == 0:
                desc_list[j] = desc_list[j] + '\n'
        desc = ' '.join(desc_list)
        img = mpimg.imread(f'{input_dir}/images/0{str(article_id)[:2]}/0{int(article_id)}.jpg')
        if len_articles//5+1 == 1:
            ax[i].imshow(img)
            ax[i].set_title(article_id)
            ax[i].set_xticks([], [])
            ax[i].set_yticks([], [])
            ax[i].grid(False)
            ax[i].set_xlabel(desc, fontsize=10)
            i += 1
        elif len_articles//5+1 > 1:
            ax[i//5, i-(i//5)*5].imshow(img)
            ax[i//5, i-(i//5)*5].set_title(article_id)
            ax[i//5, i-(i//5)*5].set_xticks([], [])
            ax[i//5, i-(i//5)*5].set_yticks([], [])
            ax[i//5, i-(i//5)*5].grid(False)
            ax[i//5, i-(i//5)*5].set_xlabel(desc, fontsize=10)
            i += 1
    plt.show()
```
