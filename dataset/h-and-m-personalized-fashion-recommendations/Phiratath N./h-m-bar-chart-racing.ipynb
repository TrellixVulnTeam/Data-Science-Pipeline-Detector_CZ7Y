```python
df = pd.read_csv('../input/h-and-m-personalized-fashion-recommendations/transactions_train.csv')
```

```python
df['t_dat']=pd.to_datetime(df['t_dat'])
```

```python
df['week']=(df['t_dat']-df['t_dat'].min()).dt.days//7
```

```python
mt_3000=df['article_id'].value_counts().loc[df['article_id'].value_counts() > 3000].index
```

```python
count_items=df[df['article_id'].isin(mt_3000)][['week','article_id','sales_channel_id']].pivot_table(index='week',columns='article_id',aggfunc='count')
```

```python
count_items=count_items['sales_channel_id']
```

```python
count_items.fillna(0,inplace=True)
```

```python
import bar_chart_race as bcr
```
