## Basic EDA follwed by some ploty express visualizations

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import gc
import os
import plotly.express as plt
from plotly.subplots import make_subplots
import plotly.graph_objects as go
gc.collect()
```

```python
trans = pd.read_csv("../input/h-and-m-personalized-fashion-recommendations/transactions_train.csv")
articles = pd.read_csv("../input/h-and-m-personalized-fashion-recommendations/articles.csv")
customers = pd.read_csv("../input/h-and-m-personalized-fashion-recommendations/customers.csv")
gc.collect()
```

```python
print("Total Transactions ",trans.shape[0] ,"Total Articles ", articles.shape[0] ,"Total Customers ", customers.shape[0])
```

```python
def missing_value(data):
    mis_data = data.isnull().sum().sort_values(ascending=False)
    per_data = ((data.isnull().sum()/data.isnull().count())*100).sort_values(ascending=False)
    #nunique_data = mis_data.nunique_data()
    ret_data = pd.concat([per_data,mis_data],axis=1,keys=["Percentage Missing","Missing Count"])
    return ret_data[ret_data["Missing Count"]>0] if ret_data[ret_data["Missing Count"]>0].shape[0]>0 else "No missing Value"

def unique_data(data):
    tot = data.count()
    nunique = data.nunique().sort_values()
    #perc = (data.nunique()/data.count()*100).sort_values(ascending=False)
    ret_data =  pd.concat([tot,nunique],keys=["Total","Unique Values"],axis=1)
    return ret_data.sort_values(by="Unique Values")
```

```python
gc.collect()
trans_artcile = pd.merge(trans,articles,on="article_id")
del trans
del articles
del customers
gc.collect()
```

## Ploting

```python
product_type_name_count = trans_artcile["product_type_name"].value_counts().reset_index()
product_type_name_count.columns = ["Product Type Name","Count"]
fig = plt.bar(product_type_name_count , x = "Product Type Name",y = "Count" ,
              barmode="group",title="Different types of Product per transactions")
fig.show()
gc.collect()
```

```python
#trans_artcile["index_group_name"].value_counts().plot(kind="bar")
index_group_name_count = trans_artcile["index_group_name"].value_counts().reset_index()
index_group_name_count.columns = ["Group Name","Count"]
fig = plt.bar(index_group_name_count , x = "Group Name",y = "Count" ,
              barmode="group",title="Different type of Group per transactions")
fig.show()
gc.collect()
```

```python

garment_group_name_count = trans_artcile["garment_group_name"].value_counts().reset_index()
garment_group_name_count.columns = ["Garment Group Name","Count"]
fig = plt.bar(garment_group_name_count , x = "Garment Group Name",y = "Count" , 
              barmode="group" , title ="Different type of garment group per Transactions")
fig.show()
gc.collect()
```

```python
section_count = trans_artcile.section_name.value_counts().reset_index()
section_count.columns = ["Section Name","Count"]
fig = plt.bar(section_count , x = "Section Name",y = "Count" , barmode="group")
fig.show()
#print(section_count)
gc.collect()
```

```python
trans_artcile["t_dat"] =  pd.to_datetime(trans_artcile["t_dat"])
gc.collect()
trans_artcile["weekday"] = trans_artcile["t_dat"].dt.weekday
gc.collect()
trans_artcile["month"] = trans_artcile["t_dat"].dt.month
gc.collect()
```

```python
section_name_month = trans_artcile.groupby(["month"])["section_name"]
gc.collect()
section_name_day = trans_artcile.groupby(["weekday"])["section_name"]
gc.collect()
```

```python
gc.collect()
name=["Jan","Feb","March","April","May","June","July","Aug","Sept","Oct","Nov","Dec"]
fig = make_subplots(3,4,subplot_titles=name,)

for i in range(12):
    month_section_name = section_name_month.get_group(i+1).reset_index()
    month_section_name =  month_section_name.section_name.value_counts().reset_index().head(10)
    month_section_name.columns = ["section_name","count"]
    fig.add_trace(go.Bar(name=name[i],x=month_section_name["section_name"],
                          y=month_section_name["count"],showlegend=False),
                  row=i//4+1,col=i%4+1)
fig.update_layout(height=1600, width=1000,title="Monthwise Sale of Different Sections over all the Transactions")
gc.collect()
fig.show()
```

```python
gc.collect()
name=["Monday","Tuesday","Wednesday","Thrusday","Friday","Saturday","Sunday"]
fig = make_subplots(2,4,subplot_titles=name,)

for i in range(7):
    day_section_name = section_name_day.get_group(i).reset_index()
    day_section_name =  day_section_name.section_name.value_counts().reset_index().head(10)
    day_section_name.columns = ["section_name","count"]
    fig.add_trace(go.Bar(name=name[i],x=day_section_name["section_name"],
                          y=day_section_name["count"],showlegend=False),
                  row=i//4+1,col=i%4+1)
fig.update_layout(height=1200, width=1000,title="Daywise Sale of Different Sections over all the Transactions")
gc.collect()
fig.show()
```
