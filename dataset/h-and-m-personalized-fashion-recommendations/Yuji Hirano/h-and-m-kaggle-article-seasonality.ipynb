```python
df_tt = pd.read_csv('/kaggle/input/h-and-m-personalized-fashion-recommendations/transactions_train.csv')
df_tt['t_month'] = df_tt['t_dat'].astype(str).str[:7]
df_tt = df_tt[['t_month', 'article_id']]
```

```python
# monthly rank
df_mon_rank = pd.DataFrame(index=[list(range(1,11))])
for month in sorted(df_tt['t_month'].unique()):
    df_temp = pd.DataFrame(np.array(df_tt.loc[df_tt['t_month']==month].value_counts('article_id')[:10].reset_index()), index=[list(range(1,11))], columns=[month, 'rank_'+str(month)])
    df_mon_rank = pd.merge(df_mon_rank, df_temp, right_index=True, left_index=True, how='left')

df_mon_rank
```

```python
import matplotlib.pyplot as plt

# #1 of the month
for col_name, item in df_mon_rank.iloc[0, ::2].iteritems():
    print(col_name+'_#1')
    img = plt.imread('/kaggle/input/h-and-m-personalized-fashion-recommendations/images/0'+str(item)[:2]+'/0'+str(item)+'.jpg')
    plt.imshow(img)

    plt.show()
```
