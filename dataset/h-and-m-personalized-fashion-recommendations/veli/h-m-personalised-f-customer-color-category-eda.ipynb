```python
df_articles=pd.read_csv('../input/h-and-m-personalized-fashion-recommendations/articles.csv')
df_transactions_train=pd.read_csv('../input/h-and-m-personalized-fashion-recommendations/transactions_train.csv')
```

```python
df_customer=pd.read_csv('../input/h-and-m-personalized-fashion-recommendations/customers.csv')
```

```python

customer_id_dic = dict(zip( df_transactions_train.article_id, df_transactions_train.customer_id))
df_articles['customer_id'] = df_articles['article_id'].map(customer_id_dic)
```

```python
customer_age_dic = dict(zip( df_customer.customer_id, df_customer.age))
df_articles['age'] = df_articles['customer_id'].map(customer_age_dic)
```

```python
df_articles=df_articles.dropna(axis=0)
```

```python
#Created New Dataset
df_age_color=df_articles[['customer_id','product_group_name' ,'colour_group_name','age']]
```

```python
df_age_color['age'].max()
```

```python
df_age_color['age'].min()
```

```python
#Divided customer ages different ranges
def age_ranges(x):
    if 24>=x>=16:
        return '16-24'
    elif 29>=x>=25 :
        return "29-25"
    elif 34>=x>=30:
        return "30-34"
    elif 39>=x>=35:
        return "35-39" 
    elif 44>=x>=40 :
        return "40-44"
    elif 49>=x>=45 :
        return "45-49"
    elif 54>=x>=50:
        return "50-54"
    elif 59>=x>=55:
        return "55-59" 
    elif 64>=x>=60 :
        return "60-64" 
    elif 69>=x>=65 :
        return "65-69"
    elif 74>=x>=70 :
        return "70-74"
    elif 79>=x>=75:
        return '75-79' 
    else:
        return "Over_79"

df_age_color['customer_age_range'] = df_age_color['age'].apply(age_ranges)
```

```python
#Count customer numbers by choice of colour and category
df_diversion=df_age_color.groupby(['product_group_name','colour_group_name','customer_age_range'],as_index=False)['customer_id'].count()
```

```python
df_diversion['customer_id'].sum()
```

```python
df_diversion['percent'] = ((df_diversion['customer_id'] / 
                  df_diversion['customer_id'].sum()) * 100)
```

```python
df_diversion
```

```python


fig = px.sunburst(df_diversion, path=['customer_age_range','product_group_name'],values='percent',width=800, height=800, 
                  title='Customer Age-Product Category EDA')

fig.update_layout(title_text='Customer Age-Product Category EDA', title_y=1,title_x=1)
fig.update_traces(textinfo='label+percent entry')
fig.update_layout(
    margin=dict(l=1, r=1, t=1, b=1)
)
fig.show()

```
