```python
import pandas as pd
import numpy as np
import random
import datetime as dt
import random
from tqdm import tqdm
```

```python
adf = pd.read_csv("../input/h-and-m-personalized-fashion-recommendations/articles.csv")
cdf = pd.read_csv("../input/h-and-m-personalized-fashion-recommendations/customers.csv")
sdf = pd.read_csv("../input/h-and-m-personalized-fashion-recommendations/sample_submission.csv")
tdf = pd.read_csv("../input/h-and-m-personalized-fashion-recommendations/transactions_train.csv")
```

```python
adf['article_id'] = adf['article_id'].astype('str')
adf['article_id'] = adf['article_id'].map(lambda x: x.zfill(10))

tdf['article_id'] = tdf['article_id'].astype('str')
tdf['article_id'] = tdf['article_id'].map(lambda x: x.zfill(10))
```

```python
tdf['t_dat'] = pd.to_datetime(tdf['t_dat'])
```

```python
tdf1 = tdf.loc[tdf['t_dat'] >= dt.datetime(2020, 9, 1)]
```

```python
tdf1 = pd.merge(tdf1, adf[['article_id', 'prod_name', 'product_type_name']], on='article_id')
```

# MAIN

```python
tdf1
```

```python
ptn_list = tdf1['product_type_name'].value_counts()[:14].index.tolist()
```

```python
ptn_dict = []
for i in tqdm(ptn_list):
    ptn_dict.extend(tdf1[tdf1['product_type_name'] == i]['prod_name'].value_counts()[:2].index.tolist())
```

```python
pn_id = []
for i in tqdm(ptn_dict):
    pn_id.append(tdf1[tdf1['prod_name'] == i]['article_id'].value_counts().index[0])
```

```python
len(pn_id)
```

```python
sub_ptn = []
for i in tqdm(range(len(sdf))):
    sub_ptn.extend(random.choices(ptn_list, k=12))
```

```python
sub_ptn[0]
```

```python
pnid_arr = np.array(pn_id).reshape(14, 2)
pnid_arr
```

```python
subm = []
for i in tqdm(sub_ptn):
    idx = ptn_list.index(i)
    subm.append(random.choice(pnid_arr[idx]))
```

```python
last_arr = np.array(subm).reshape(1371980, 12)
```

```python
submission = []
for i in tqdm(last_arr):
    submission.append(" ".join(i))
```

```python
sdf['prediction'] = submission
```

```python
sdf.to_csv("submission.csv", index=False)
```
