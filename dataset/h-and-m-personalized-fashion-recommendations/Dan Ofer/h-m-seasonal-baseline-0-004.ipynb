# H&M - Baseline: 12 most popular SEASONAL items

### A naive approach system using the 12 items that were "most popular" for the specific month!

* Test period: 7 days, From 2020-09-22 to 2020-09-29 

* V1: september only, most popular by total users - 0.004 public LB
* V2: most popular by total amount - 0.000 public LB
* V3: september+October only, most popular by total users

Normally this approach would be VERY strong as a baseline - but! COVID messes this sort of consumer behavior up, especially given that out september history was during lockdowns. We lack relevant historical data!

```python
import pandas as pd
```

```python
DATA_PATH = "../input/h-and-m-personalized-fashion-recommendations/"
```

```python
df = pd.read_csv(DATA_PATH+'transactions_train.csv', dtype={'article_id': str},
                parse_dates=['t_dat'],infer_datetime_format=True,usecols=['t_dat', 'customer_id', 'article_id'])
df_sub = pd.read_csv(DATA_PATH+'sample_submission.csv')

articles = pd.read_csv(DATA_PATH+'articles.csv').set_index('article_id')
df.head()
```

### Keep only records for September
* Could try also +- a month - winter/autumn season

```python
print(df.shape[0])
df = df.loc[df['t_dat'].dt.month==9]
# df = df.loc[df['t_dat'].dt.month.between(9,10)] ## september through october?
print(df.shape[0])
df.head()
```

# Get the 12 most popular items
* By total users OR by total purchases


```python
## 12 by unique users:
print("top 12 by unique customers")
# top_12_items = df.groupby('article_id')['customer_id'].nunique().sort_values(ascending=False).head(12).index#.tolist() # orig
top_12_byUser_items = df.groupby('article_id')['customer_id'].nunique().sort_values(ascending=False).head(12).index.astype(int)

display(articles.loc[top_12_byUser_items])

# top_12_byUser_items = top_12_byUser_items.astype(str).tolist()

top_12_byUser_items= df.groupby('article_id')['customer_id'].nunique().sort_values(ascending=False).head(12).index.tolist() # recalc
print(top_12_byUser_items)
```

```python
print("top 12 by overall popularity")
# top_12_items = df['article_id'].value_counts().sort_values(ascending=False).head(12).index.tolist()
top_12_pop_items = df['article_id'].value_counts().sort_values(ascending=False).head(12).index.astype(int)

display(articles.loc[top_12_pop_items])

# top_12_pop_items = top_12_pop_items.astype(str).tolist()
top_12_pop_items = df['article_id'].value_counts().sort_values(ascending=False).head(12).index.tolist()
top_12_pop_items
```

```python
top_12_items = top_12_byUser_items 
# top_12_items = top_12_pop_items # inferior
```

# Submit

```python
# Predict for everyone
df_sub['prediction'] =  ' '.join(top_12_items)
df_sub.to_csv('submission.csv', index=False)
df_sub.head()
```
