```python
# import seaborn
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
plt.figure(figsize = (16,8))
plt.title("Histogram of data in Activity Dataframe")
sns.countplot(hue = 'index_code', x='department_name',data = article);
plt.show()
```

```python
f, ax = plt.subplots(figsize=(10, 5))
ax = sns.histplot(data=article, y='index_name', color='green')
ax.set_xlabel('count by index name')
ax.set_ylabel('index name')
plt.show()
```

```python
f, ax = plt.subplots(figsize=(10, 5))
ax = sns.histplot(data=article, y='garment_group_name', color='orange', hue='index_group_name', multiple="stack")
ax.set_xlabel('count by garment group')
ax.set_ylabel('garment group')
plt.show()
```

```python
sns.set_style("darkgrid")
f, ax = plt.subplots(figsize=(10,5))
ax = sns.histplot(data=customer, x='age', bins=50, color='blue')
ax.set_xlabel('Distribution of the customers age')
plt.show()
```

```python
sns.set_style("darkgrid")
f, ax = plt.subplots(figsize=(10,5))
ax = sns.histplot(data=customer, x='club_member_status', color='Red')
ax.set_xlabel('Distribution of club member status')
plt.show()
```
