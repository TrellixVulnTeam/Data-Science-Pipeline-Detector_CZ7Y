```python
temp = a.groupby(["garment_group_name"])["product_type_name"].nunique()
df = pd.DataFrame({"Garment Group": temp.index, "Product Types": temp.values})
df = df.sort_values(["Product Types"], ascending=False)
plt.figure(figsize=(16, 5))
plt.title("Number of Product Types each Garment Group")
s = sns.barplot(x="Garment Group", y="Product Types", data=df, palette="cubehelix")
s.set_xticklabels(s.get_xticklabels(), rotation=90)
locs, labels = plt.xticks()
plt.show()
```

```python
temp = c.groupby(["t_dat"])["price"].nunique()
df = pd.DataFrame({"t_dat": temp.index, "price": temp.values})
df = df.sort_values(["price"], ascending=False)
plt.figure(figsize=(20, 10))
plt.title("prices at different dates")
s = sns.barplot(x="t_dat", y="price", data=df, palette="cubehelix")
s.set_xticklabels(s.get_xticklabels(), rotation=90)
locs, labels = plt.xticks()
plt.show()
```
