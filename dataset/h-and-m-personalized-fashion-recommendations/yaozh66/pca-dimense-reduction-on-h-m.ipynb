```python
import numpy as np
import pandas as pd
image_embeddings=pd.read_feather("../input/embeddings/embedding_image.feature")
image_embedding_value=image_embeddings.iloc[:,1:513].values
```

```python
image_embeddings
```

```python
text_embeddings=pd.read_feather("../input/embeddings/embedding_text.feature")
text_embedding_value=text_embeddings.iloc[:,1:768].values
```

```python
text_embedding_value
```

```python
from sklearn.decomposition import PCA
n_dimension=[25,50,100,150,200,250,300,450]
from tqdm import tqdm
for dimension in tqdm(n_dimension):
    pca = PCA(n_components=dimension)
    transformed_image=pca.fit_transform(image_embedding_value)
    pca = PCA(n_components=dimension)
    transformed_text=pca.fit_transform(text_embedding_value)
    stack=np.column_stack((transformed_image,transformed_text))
    embeddings=pd.DataFrame(stack,columns=["embedding"+str(i+1) for i in range(2*dimension)])
    embeddings.to_feather("./embeddings"+str(dimension)+".feather")
```
