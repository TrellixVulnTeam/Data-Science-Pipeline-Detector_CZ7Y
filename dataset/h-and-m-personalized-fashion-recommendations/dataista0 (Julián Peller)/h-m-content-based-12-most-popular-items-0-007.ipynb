## A content-based naive recommendation system using the 12 items that were purchased the most ("most popular") in the full transaction history for the competition [H&M Personalized Fashion Recommendations](https://www.kaggle.com/c/h-and-m-personalized-fashion-recommendations).

# Read dataframes

````python
import pandas as pd
df = pd.read_csv('../input/h-and-m-personalized-fashion-recommendations/transactions_train.csv', dtype={'article_id': str})
df_sub = pd.read_csv('../input/h-and-m-personalized-fashion-recommendations/sample_submission.csv')
df.head()
````

````python
df.t_dat.max()
````

# Get the 12 most popular items in the last month

https://www.kaggle.com/hengzheng/time-is-our-best-friend

````python
top_12_items = df[df['t_dat'] > '2020-08-21'].groupby('article_id')['customer_id'].nunique().sort_values(ascending=False).head(12).index.tolist()
top_12_items
````

# Submit

````python
# Predict the same for everyone
df_sub['prediction'] =  ' '.join(top_12_items)
df_sub.to_csv('submission.csv', index=False)
df_sub.head()
````

### Condensed in 5 lines of code:

```python
import pandas as pd
df = pd.read_csv('../input/h-and-m-personalized-fashion-recommendations/transactions_train.csv', dtype={'article_id': str})
df_sub = pd.read_csv('../input/h-and-m-personalized-fashion-recommendations/sample_submission.csv')
df_sub['prediction'] =  ' '.join(df[df['t_dat'] > '2020-08-21'].groupby('article_id')['customer_id'].count().sort_values(ascending=False).head(12).index.tolist())
df_sub.to_csv('submission.csv', index=False)
```
