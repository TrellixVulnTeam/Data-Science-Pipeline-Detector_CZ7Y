This notebook creates two new datasets
First one has customer, article and rating columns with rating column representing the number of times the customer purchased the item.
The second one has all items each customer purchased ordered chronologically.

```python
import numpy as np # linear algebr
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
col_list = ['customer_id', 'article_id', 'sales_channel_id']
df = pd.read_csv('../input/h-and-m-personalized-fashion-recommendations/transactions_train.csv', usecols=col_list, dtype=str)
```

```python
ratings = df.groupby(['customer_id', 'article_id']).count()
```

```python
df2 = pd.DataFrame(tuple(ratings.index.values))
df2['rating'] = ratings.sales_channel_id.values
df2.columns = ['customer_id', 'article_id', 'rating']
```

```python
df2.to_csv('ratings.csv')
```

```python
del df2
```

```python
seq = df[['customer_id', 'article_id']].groupby('customer_id')
```

```python
customers = []
articles = []
for group in seq.groups:
    customers.append(group)
    articles.append(seq.get_group(group).article_id.values.tolist())
```

```python
seq_df = pd.DataFrame({'customer':customers, 'sequence':articles})
```

```python
seq_df.to_csv('customer_sequence.csv')
```
