```python
#Importing libraries
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
from PIL import Image

# Uploading data
articles = pd.read_csv('../input/h-and-m-personalized-fashion-recommendations/articles.csv')
customers = pd.read_csv('../input/h-and-m-personalized-fashion-recommendations/customers.csv')
transactions_train = pd.read_csv('../input/h-and-m-personalized-fashion-recommendations/transactions_train.csv')
```

<h1 style="color:red;">Top Selling Items - Valentine Week(7Feb -14Feb)</h1>

```python
# colour_group_name
transactions_train['t_dat'] = pd.to_datetime(transactions_train['t_dat'], infer_datetime_format=True)
transactions_train['year'] = transactions_train['t_dat'].dt.year
transactions_train['mon'] = transactions_train['t_dat'].dt.month
transactions_train['day'] = transactions_train['t_dat'].dt.day

a = transactions_train[(transactions_train['mon'] == 2) & (transactions_train['day'] >=7) & (transactions_train['day'] <=14 )][['article_id', 'year']]
b = articles[['article_id', 'colour_group_name']]
val = pd.merge(a, b, how = 'inner', on = ['article_id']).reset_index()
c = val.copy()
c = c.groupby(by = ['article_id']).agg({'article_id':'count'})
c.columns = ['Count']
c = c.reset_index()
c = c.sort_values(by = ['Count'], ascending=[False])
c = c.head(11)

plt.figure(figsize=(28, 10))
i = 1
for j,x in enumerate(c['article_id'].to_list()):
    try:
        image = Image.open("../input/h-and-m-personalized-fashion-recommendations/images/0"+str(x)[:2]+"/0"+str(x)+".jpg")
        plt.subplot(2, 5, i)
        plt.imshow(image)
        #plt.axis('off')
        plt.title(x)
        plt.xlabel("Sell Count = "+str(c['Count'].to_list()[j]))
        i+=1
    except:
        pass
```

<h1 style="color:red;">Let do Top selling Color Analysis</h1>

```python
c = val.copy()
c['final_color'] = None
c['final_color'] = np.where(c['colour_group_name'].str.lower().str.contains("red"), "red", c['final_color'])
c['final_color'] = np.where(c['colour_group_name'].str.lower().str.contains("green"), "green", c['final_color'])
c['final_color'] = np.where(c['colour_group_name'].str.lower().str.contains("white"), "white", c['final_color'])
c['final_color'] = np.where(c['colour_group_name'].str.lower().str.contains("pink"), "pink", c['final_color'])
c['final_color'] = np.where(c['colour_group_name'].str.lower().str.contains("blue"), "blue", c['final_color'])
c['final_color'] = np.where(c['colour_group_name'].str.lower().str.contains("yellow"), "yellow", c['final_color'])
c['final_color'] = np.where(c['colour_group_name'].str.lower().str.contains("black"), "black", c['final_color'])
c['final_color'] = np.where(c['colour_group_name'].str.lower().str.contains("orange"), "orange", c['final_color'])
c = c[c['final_color'].isna() == False]

import plotly.express as px
d = c[c['year'] == 2019]['final_color'].value_counts().reset_index()
d.columns = ['colour', 'cnt']
#Creating grpah
colours = {
    "black": "#000000",
    "blue": "#0000FF",
    "white": "#FFFFFF",
    "red": "#FF0000",
    "pink": "#FF00FF",
    "green": "#008000",
    "yellow": "#FFFF00",
    "orange": "#FFA500"
}

plotly_plt = px.bar(d, x="colour", y=["cnt"], 
             #barmode = "group",
#              labels={'color':'Color',
#                      'cnt':'Count'                    },
             title= "2019 Valentine Week - Top Selling Color",
             #color=["default_pct", "total_pct"],
             #hover_name = "default_pct",
             #text = ["default_pct", "total_pct"],
             color="colour",
             text="value",
             width=1000, height=500,
             color_discrete_map=colours,
            )
plotly_plt.show()

d = c[c['year'] == 2020]['final_color'].value_counts().reset_index()
d.columns = ['colour', 'cnt']
plotly_plt = px.bar(d, x="colour", y=["cnt"], 
             #barmode = "group",
#              labels={'color':'Color',
#                      'cnt':'Count'                    },
             title= "2020 Valentine Week - Top Selling Color",
             #color=["default_pct", "total_pct"],
             #hover_name = "default_pct",
             #text = ["default_pct", "total_pct"],
             color="colour",
             text="value",
             width=1000, height=500,
             color_discrete_map=colours,
            )
plotly_plt.show()
```

```python
def color_analysis(inp_color, no=5):
    c = val.copy()
    c = c[c['colour_group_name'].str.lower().str.contains(inp_color.lower())]
    c = c.groupby(by = ['article_id', 'colour_group_name']).agg({'article_id':'count'})
    c.columns = ['Count']
    c = c.reset_index()
    c = c.sort_values(by = ['Count'], ascending=[False])
    c = c.head(no)

    plt.figure(figsize=(28, 10))
    i = 1
    for j,x in enumerate(c['article_id'].to_list()):
        try:
            image = Image.open("../input/h-and-m-personalized-fashion-recommendations/images/0"+str(x)[:2]+"/0"+str(x)+".jpg")
            plt.subplot(2, 5, i)
            plt.imshow(image)
            #plt.axis('off')
            plt.title(x)
            plt.xlabel("Sell Count = "+str(c['Count'].to_list()[j]))
            plt.xlabel("Color = "+str(c['colour_group_name'].to_list()[j]))
            i+=1
        except:
            pass
```
