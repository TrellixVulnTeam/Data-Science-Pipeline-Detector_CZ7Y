## Helper function that allows to print the distribuction of custumers age with respect to some other variable

```python
def ageDistributionPlot(df, legend):
    fig, ax = plt.subplots(figsize=(10,5))
    ax = sns.histplot(data=df, x='age', bins=df['age'].nunique(), color='orange', stat="percent")
    ax.set_xlabel(legend)
    for loc in ['bottom', 'left']:
        ax.spines[loc].set_visible(True)
        ax.spines[loc].set_linewidth(2)
        ax.spines[loc].set_color('black')
    ax.text(12, 5.5, legend , color='black', fontsize=10, ha='left', va='bottom', weight='bold', style='italic')
    plt.show()
```
