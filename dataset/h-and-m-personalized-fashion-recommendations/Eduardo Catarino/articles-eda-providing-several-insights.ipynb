# Helper Functions

```python
#Function cat_sub
# col_names - Array containing the column names from witch we would like to discriminate the categories and subcategories
# art_df - Dataframe containing the dataset
# art_col_values - String containing the name of the column from witch we can count the values (Default value = 'articles_id')
# returns: df - Dataframe with MultiIndexed by col_names
# 
# In order to print the results use the return value as folows:
#     dfTest = cat_sub(['product_group_name','product_type_name'],articles_df)
#     dfTest.groupby(cols).count()
def cat_sub(col_names, art_df, art_col_values = 'article_id'):

    aprod = art_df[col_names]
    index = pd.MultiIndex.from_frame(aprod, names=col_names)
    id = art_df[[art_col_values]]
    df = pd.DataFrame(id.values.tolist(), index=index)
    df.groupby(col_names).count()
    return df
```
