# Calculate Association Rules

```python
from mlxtend.frequent_patterns import association_rules
```

```python
is_ap
```

```python
rules = association_rules(is_ap, metric="lift")
rules.head()
```

```python
def predict(antecedent, rules, max_results= 6):
    
    # get the rules for this antecedent
    preds = rules[rules['antecedents'].astype(str).str.contains(antecedent)]
    
    # a way to convert a frozen set with one element to string
    preds = preds['consequents'].apply(iter).apply(next)
    
    return preds[:max_results]

algo_dict = {}

for i in rules['antecedents']:
    
    #Convert our algorithm to key pairs
    antecedents = list(i)[0]
    consequent_ls = predict(str(antecedents), rules)
    
    for j in consequent_ls:
        algo_dict[antecedents] = j
    
    
algo_dict
```

```python
#Predictions based on apriori
#Why is it just one prediction?
preds = predict('158340001', rules)
preds
```

```python
#local_cv = True

import datetime
from tqdm import tqdm

def apk(actual, predicted, k=12):
    if len(predicted)>k:
        predicted = predicted[:k]

    score = 0.0
    num_hits = 0.0

    for i,p in enumerate(predicted):
        if p in actual and p not in predicted[:i]:
            num_hits += 1.0
            score += num_hits / (i+1.0)

    if not actual:
        return 0.0

    return score / min(len(actual), k)

def mapk(actual, predicted, k=12):
    return np.mean([apk(a,p,k) for a,p in zip(actual, predicted)])
```

```python
data = pd.read_csv("../input/h-and-m-personalized-fashion-recommendations/transactions_train.csv", dtype={'article_id':str})
data["t_dat"] = pd.to_datetime(data["t_dat"])
data.head()

val = data.loc[data["t_dat"] >= datetime.datetime(2020,9,16)]

positive_items_val = val.groupby(['customer_id'])['article_id'].apply(list)
val_users = positive_items_val.keys()
```

```python
val_items = []
pred_items = []

for i,user in tqdm(enumerate(val_users)):
    val_items.append(positive_items_val[user])
    pred_items.append(preds[preds['customer_id'] == user]['prediction'].tolist())
    
print("Total users in validation:", len(val_users))
```

```python
print("mAP Score on Validation set:", mapk(val_items, pred_items))
```
