```python
import pickle

import pandas as pd
import numpy as np
```

```python
def pickle_variables(variable_names, delete_variables=False):
    """
    saves to current directory
    if delete_variables=True, will also delete variables

    pickle variables with format <variable_name>.pkl
    """
    for variable_name in variable_names:
        with open(f"{variable_name}.pkl", "wb") as f:
            pickle.dump(globals()[variable_name], f)
        if delete_variables:
            del globals()[variable_name]
```

```python
%%time
transactions_train = pd.read_csv("/kaggle/input/h-and-m-personalized-fashion-recommendations/transactions_train.csv",
                           dtype={"t_dat": "object", "customer_id": "object", "article_id": "object", "price": float, "sales_channel_id": int})
customers = pd.read_csv("/kaggle/input/h-and-m-personalized-fashion-recommendations/customers.csv")
articles = pd.read_csv("/kaggle/input/h-and-m-personalized-fashion-recommendations/articles.csv", dtype={"article_id": "object"})
```

```python
pickle_variables(["transactions_train", "customers", "articles"])
```

```python
customers_50pct = customers.sample(frac=0.5, replace=False).reset_index(drop=True)
customers_50pct_ids = set(customers_50pct["customer_id"])
transactions_train_50pct = transactions_train.query("customer_id in @customers_50pct_ids").reset_index(drop=True)
pickle_variables(["transactions_train_50pct", "customers_50pct"], delete_variables=True)
```

```python
customers_5pct = customers.sample(frac=0.05, replace=False).reset_index(drop=True)
customers_5pct_ids = set(customers_5pct["customer_id"])
transactions_train_5pct = transactions_train.query("customer_id in @customers_5pct_ids").reset_index(drop=True)
pickle_variables(["transactions_train_5pct", "customers_5pct"], delete_variables=True)
```
