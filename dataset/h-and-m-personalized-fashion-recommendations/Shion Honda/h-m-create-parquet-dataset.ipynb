```python
import pandas as pd
import numpy as np
```

```python
%%time
transactions_train = pd.read_csv("/kaggle/input/h-and-m-personalized-fashion-recommendations/transactions_train.csv",
                           dtype={"t_dat": "object", "customer_id": "object", "article_id": "object", "price": float, "sales_channel_id": int})
customers = pd.read_csv("/kaggle/input/h-and-m-personalized-fashion-recommendations/customers.csv")
articles = pd.read_csv("/kaggle/input/h-and-m-personalized-fashion-recommendations/articles.csv", dtype={"article_id": "object"})
```

```python
transactions_train.to_parquet('transactions_train.parquet')
customers.to_parquet('customers.parquet')
articles.to_parquet('articles.parquet')
```

```python
customers_50pct = customers.sample(frac=0.5, replace=False)
customers_50pct_ids = set(customers_50pct["customer_id"])
transactions_train_50pct = transactions_train.query("customer_id in @customers_50pct_ids")

transactions_train_50pct.to_parquet('transactions_train_50pct.parquet')
customers_50pct.to_parquet('customers_50pct.parquet')
```

```python
customers_5pct = customers.sample(frac=0.05, replace=False)
customers_5pct_ids = set(customers_5pct["customer_id"])
transactions_train_5pct = transactions_train.query("customer_id in @customers_5pct_ids")

transactions_train_5pct.to_parquet('transactions_train_5pct.parquet')
customers_5pct.to_parquet('customers_5pct.parquet')
```
