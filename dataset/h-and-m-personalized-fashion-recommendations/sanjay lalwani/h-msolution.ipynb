```python
def missing_percent_of_column(train_set):
    nan_percent = 100*(train_set.isnull().sum()/len(train_set))
    nan_percent = nan_percent[nan_percent>0].sort_values(ascending=False).round(1)
    DataFrame = pd.DataFrame(nan_percent)
    # Rename the columns
    mis_percent_table = DataFrame.rename(columns = {0 : '% of Misiing Values'}) 
    # Sort the table by percentage of missing descending
    mis_percent = mis_percent_table
    return mis_percent
```

# Data Analysis

**Exploratory data analysis**

Lets first visualise data with the help of different graphs, this would help us to understand data the data.

```python
f, ax = plt.subplots(figsize=(10,5))
sns.histplot(x="age", data=cust, bins=50, kde=True)
ax.set_xlabel('Distribution of the customers age')
plt.show()
```

```python
f, ax = plt.subplots(figsize=(7,5))
sns.countplot(data=cust, x="club_member_status")
plt.show()
```

```python
f, ax = plt.subplots(figsize=(7,5))
sns.countplot(data=cust, x="fashion_news_frequency")
plt.show()
```

```python
f, ax = plt.subplots(figsize=(10,5))
ax = sns.boxplot(data=train, x='price')
ax.set_xlabel('Price outliers')
plt.show()
```

```python
f, ax = plt.subplots(figsize=(15,5))
pgn = artc.product_group_name.value_counts().rename_axis('product_group_name').reset_index(name='counts')
sns.barplot(data=pgn, x="counts", y = 'product_group_name',palette="Blues_d")
ax.set_xlabel('Count of article by product_group_name')
plt.show()
```

```python
f, ax = plt.subplots(figsize=(10,5))
sns.countplot(data=artc, y="index_name",dodge=True)
plt.show()
```

```python
f, ax = plt.subplots(figsize=(10,6))
sns.histplot(data=artc, y="index_group_name",hue='index_name', multiple="stack")
plt.show()
```
