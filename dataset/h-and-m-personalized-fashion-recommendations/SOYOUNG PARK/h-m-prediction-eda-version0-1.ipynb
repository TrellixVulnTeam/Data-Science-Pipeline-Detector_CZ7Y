**<Data Explanation>**
 
* articles.csv - detailed metadata for each article_id available for purchase
* customers.csv - metadata for each customer_id in dataset
* sample_submission.csv - a sample submission file in the correct format
* transactions_train.csv - the training data, consisting of the purchases each customer for each date, as well as additional information. Duplicate rows correspond to multiple purchases of the same item. 
* Your task is to predict the article_ids each customer will purchase during the 7-day period immediately after the training data period.
*     
NOTE: You must make predictions for all customer_id values found in the sample submission. All customers who made purchases during the test period are scored, regardless of whether they had purchase history in the training data.

```python
#We can notice similarity between article_id and product_code. And from the numbers of values, it indicates that there are certian main categories of the product
articles['product_code'].value_counts()
```

```python
main_prod_df = articles.groupby( ['product_code','product_type_no','product_type_name','prod_name'] ).size().to_frame(name = 'count').reset_index()
main_prod_df.sort_values(['count'],ascending=False)
```

```python
main_prod_df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,4))
plt.title('Number of articles: Product code')
plt.xlabel('Product code Range')
plt.ticklabel_format(axis = 'x',style ='plain')
sns.lineplot(data =main_prod_df,x= 'product_code',y='count')
# It seems like Product Code range between 400000 and 800000 has lots of kinds of article.
```

```python
articles[(articles['product_code']>=600000)&(articles['product_code']<800000)].sample(10).sort_values(by='product_type_no')
#It may not be accurate, but we can assume that product code with lots of kinds of article would be one of the common kinds of clothes type.
```

```python
pd.pivot_table(articles,index =['product_group_name','product_type_name','product_type_no'],values = ['article_id'],aggfunc =['count','min','max'])                     
#This analysis indicates that there are certain product_type_no depending on product_group_name and product_type_name feature which means that we can discover what kind of product it is only by product_type_no!
#Also, the range of article_id varies regardless of product type. 
```
