```python
articles= pd.read_csv("../input/h-and-m-personalized-fashion-recommendations/articles.csv")
articles
```

```python
for col in articles.columns:
    print(col)
```

```python
customers= pd.read_csv("../input/h-and-m-personalized-fashion-recommendations/customers.csv")
customers
```

```python
transaction= pd.read_csv("../input/h-and-m-personalized-fashion-recommendations/transactions_train.csv",nrows=100000)
transaction
```

```python
submission= pd.read_csv("../input/h-and-m-personalized-fashion-recommendations/sample_submission.csv")
submission
```

```python
df_train=pd.merge(transaction, articles, on="article_id", how="left")
df_train
```

```python
for col in df_train.columns:
    print(col)
```

```python
df_train['prod_name'].value_counts()
type(df_train['prod_name'].value_counts())
df_article = df_train['prod_name'].value_counts().reset_index()
df_article.columns = ['Article', 'Count']
df_article
```

```python
df_train1 = df_train.groupby(['customer_id','article_id'])['article_id'].sum()
df_train1
```

```python
df_train2 = df_train.groupby(['article_id','customer_id'])['article_id'].sum()
print(df_train2)
```

```python
df_train_merge=pd.merge(df_train2,df_train1, on="customer_id", how = "right")
df_train_merge_dataframe = pd.DataFrame(df_train_merge)
df_train_merge_dataframe
```

```python
df_train_final = df_train_merge_dataframe.drop(columns=['article_id_y'])
df_train_final
```

```python
submission = df_train_final.groupby('customer_id')['article_id_x'].agg(list)
submission.shape[0]
```

```python
df_submission = pd.DataFrame(submission)
df_submission_renamed = df_submission.rename(columns = {'article_id_x':'prediction'})
df_submission_reindexed = df_submission_renamed.reset_index()
df_submission_reindexed['prediction'] = df_submission_reindexed['prediction'].astype(str).str.replace(r'\[|\]|,', '')
df_submission_reindexed
```

```python
df_submission=pd.merge(df_submission_reindexed,customers, on="customer_id", how = "right")
df_submission
```

```python
df_submission_final = df_submission.drop(columns=['FN','Active','club_member_status','fashion_news_frequency','age','postal_code'])
df_submission_final
```

```python
df_submission_final.to_csv("submission.csv", index=False)
```
