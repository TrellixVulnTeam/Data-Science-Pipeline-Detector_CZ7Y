```python
def getImagePaths(path):
    image_names = []
    for dirname, _, filenames in os.walk(path): #　path⇒それぞれの画像ファイル walk⇒ファイル内のディレクトリ名とファイル名を取り出している。
        for filename in filenames:
            fullpath = os.path.join(dirname, filename) # 一つ一つのファイル名とディレクトリ名を一つのパスにしている。
            image_names.append(fullpath)
    return image_names #パスとしての表示としている

def display_multiple_img(images_paths, rows, cols):
    figure, ax = plt.subplots(nrows=rows,ncols=cols,figsize=(16,8) ) #与えたデータ分、サブプロットにしている。
    for ind,image_path in enumerate(images_paths):  #インデックス番号と画像のパスを同時に出している。
        image=cv2.imread(image_path) #cvで画像の読み込み
        
        try:
            ax.ravel()[ind].imshow(image)
            ax.ravel()[ind].set_axis_off() #Axes.set_axis_off()[source]X 軸と Y 軸をオフにします。
        except:
            continue;
    plt.tight_layout() #tight_layout() メソッドはサブプロット間の正しい間隔を自動的に維持します。
    plt.show()
```

```python
images_path = getImagePaths(images_dir)
display_multiple_img(images_path[0:30], 6, 5) #30個までのデータを6行5列で表している
```
