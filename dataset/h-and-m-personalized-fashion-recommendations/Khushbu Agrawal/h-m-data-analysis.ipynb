```python
df_tran_arti = pd.merge(
    df_tran, 
    df_articles[["article_id", "index_group_name", "product_type_name", "graphical_appearance_name", "colour_group_name"]], 
    how = "left", 
    on = "article_id"
)
```

```python
df_tran_arti.columns
```

# Checking sales of product types

```python
ax = sns.countplot(x="product_type_name", data = df_tran_arti)
a = plt.xticks(rotation=90)
```

As we are interseted in september last week, lets check sales of items from 15 sept to 15 octomber for each year

```python
df_tran_arti.t_dat = pd.to_datetime(df_tran_arti.t_dat)
```

```python
a = df_tran_arti.loc[((df_tran_arti["t_dat"] > "2018-09-15") & (df_tran_arti["t_dat"] < "2018-10-14")) |
                    ((df_tran_arti["t_dat"] > "2019-09-15") & (df_tran_arti["t_dat"] < "2019-10-14")) |
                    ((df_tran_arti["t_dat"] > "2020-09-15") & (df_tran_arti["t_dat"] < "2020-10-14"))]
```

```python
ax = sns.countplot(x="product_type_name", data = a)
a = plt.xticks(rotation=90)
```

As this is for september months, sales of sweater is quite high whereas Dresses, bikini top and swimwears and shorts is quite low.

# Sales based on age

```python
df_tran_arti_age = pd.merge(
    df_tran_arti, 
    df_cust[["customer_id", "age"]], 
    how = "left", 
    on = "customer_id"
)
df_tran_arti_age["age1"] =  np.floor(df_tran_arti_age["age"] / 10) * 10
temp = df_tran_arti_age.groupby(["age1", 'index_group_name']).count()["t_dat"].unstack(1).replace(np.nan, 0)
```

```python
sns.heatmap(temp, cmap="PiYG")
```
