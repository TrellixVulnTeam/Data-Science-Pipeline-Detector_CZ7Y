This is the simplest "dumb" algorithm I could think of where I recommend the most popular 12 products in the last month to all users.

```python
import pathlib

import numpy as np
import pandas as pd
```

```python
DATA = pathlib.Path("/kaggle/input/h-and-m-personalized-fashion-recommendations/")

# articles = pd.read_csv(DATA / "articles.csv")
# customers = pd.read_csv(DATA / "customers.csv")
transactions = pd.read_csv(DATA / "transactions_train.csv")
transactions["t_dat"] = pd.to_datetime(transactions["t_dat"])
submission = pd.read_csv(DATA / "sample_submission.csv")
```

```python
counts = transactions["article_id"].value_counts()
pop_items = counts[:12]
pop_items
```

```python
latest_pop_items = transactions[transactions["t_dat"] >= "2020-09-01"]["article_id"].value_counts()[:12]
latest_pop_items
```

```python
submission["prediction"] = " ".join([str(item_id) for item_id in latest_pop_items.index.tolist()])
submission.to_csv("/kaggle/working/submission.csv", index=False)
```
