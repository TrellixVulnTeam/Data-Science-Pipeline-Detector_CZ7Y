# **<span id="Customers" style="color:#7AB495;">Customers</span>**

```python
customers.columns
```

**FN - 1 if a customer get Fashion News newsletter**

**Active - 1 if the customer is active for communication**

```python
customers['FN'].unique(), customers['Active'].unique()
```

**Club membership status**

```python
customers['club_member_status'].unique()
```

```python
customers[customers['club_member_status'].isna()]
```

nan -> 'None'

```python
customers.loc[customers['club_member_status'] == np.nan, 'club_member_status'] = 'None'
```

**News frequency**

```python
customers['fashion_news_frequency'].unique()
```

**NONE, None and nan. 3 values 1 meaning - fix it**

```python
customers.loc[customers['fashion_news_frequency'].isin(['NONE', 'None', np.nan]), 'fashion_news_frequency'] = 'None'
customers['fashion_news_frequency'].unique()
```

**Age**

```python
sns.set_style("darkgrid")
f, ax = plt.subplots(figsize=(10,7))
ax = sns.histplot(data=customers, x='age', bins=50, color='green')
ax.set_xlabel('Distribution of the age')
plt.show()
```

```python
customers['fashion_news_frequency'].index
```

**Distribution of news frequency**

```python
pie_data = customers.groupby('fashion_news_frequency')['customer_id'].count()
pie_data
```

```python
f, ax = plt.subplots(figsize=(10,7))
colors = sns.color_palette('deep')
ax.pie(pie_data, labels=pie_data.index, colors = colors)
ax.set_facecolor('lightgrey')
ax.set_xlabel('Distribution of news frequency')
plt.show()
```

## Price

```python
f, ax = plt.subplots(figsize = (10, 7))
plt.xlim(0, 0.3)
ax = sns.histplot(data = df, x = 'price', bins = 75, color = 'green')
ax.set_xlabel('Distribution of price')
plt.show()
```

```python
df['price'].describe()
```

```python
f, ax = plt.subplots(figsize = (12,8))
ax = sns.boxplot(data = df, x = 'price', color = "green")
ax.set_xlabel('Price boxplot')
plt.show()
```

### Sales_channel_id

```python
pie_data = df.groupby('sales_channel_id').count()['customer_id']
pie_data
```

```python
f, ax = plt.subplots(figsize = (10, 7))
colors = sns.color_palette('deep')
ax.pie(pie_data, labels=['Online', 'Offline'], colors = colors)
ax.set_facecolor('lightgrey')
ax.set_xlabel('Distribution of sales channel')
plt.show()
```
