# 4. Plot some Products

```python
def plot_images(Path):
    images = []
    for img_path in glob.glob(Path + '/*.jpg'):
        images.append(mpimg.imread(img_path))

    plt.figure(figsize=(20,20))
    columns = 3
    for i, image in enumerate(images):
        plt.subplot(len(images) // columns + 1, columns, i + 1)
        plt.imshow(image)
```

```python
plot_images('../input/h-and-m-personalized-fashion-recommendations/images/010')
```

```python
plot_images('../input/h-and-m-personalized-fashion-recommendations/images/015')
```
