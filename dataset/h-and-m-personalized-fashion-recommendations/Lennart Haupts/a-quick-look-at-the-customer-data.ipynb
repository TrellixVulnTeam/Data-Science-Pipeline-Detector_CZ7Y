```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
```

```python
customers = pd.read_csv('../input/h-and-m-personalized-fashion-recommendations/customers.csv')
```

```python
customers.shape
```

```python
# Missing values
customers.isna().sum()
```

```python
# Unique values
for c in customers.columns:
    print(f'{c}: {len(customers[c].unique())}')
```

```python
# Value counts for FN
customers.FN.value_counts()
```

```python
# Value counts for Active
customers.Active.value_counts()
```

```python
# Filling missing values for FN and Active with 0
customers.FN = customers.FN.fillna(0)
customers.Active = customers.Active.fillna(0)
```

```python
# Values counts for membership status
customers.club_member_status.value_counts()
```

```python
# Value counts for frequency of fashion news
customers.fashion_news_frequency.value_counts()
```

```python
# converting fashion news to all lower to get rid of NONE and None
customers.fashion_news_frequency = customers.fashion_news_frequency.str.lower() 
```

```python
# Count plots for FN and Active
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,7))
sns.countplot(ax=ax1, x='FN', data=customers)
sns.countplot(ax=ax2, x='Active', data=customers)
plt.show()
```

```python
# linear relationship between FN and Active
pearsonr(customers.FN, customers.Active)
```

```python
print('Fn but not Active: ' +  str(len(customers[(customers.FN == 1) & (customers.Active == 0)])))
print('Acitve but not FN:     ' +  str(len(customers[(customers.FN == 0) & (customers.Active == 1)])))
```

```python
# countplot frequency of fashion news
fig = plt.figure(figsize=(7, 7))
sns.countplot(x='fashion_news_frequency', data=customers)
plt.show()
```

```python
# age distribution
fig = plt.figure(figsize=(7, 7))
sns.histplot(customers.age, bins=80, kde = True)
plt.show()
```

```python
# Value counts of the top 40 most common postal codes
customers.postal_code.value_counts()[:40]
```

```python
fig = plt.figure(figsize=(20, 7))
sns.histplot(customers.postal_code.value_counts()[1:], bins=250, kde=False)
plt.xlim(0, 50)
plt.tight_layout()
plt.show()
```
