## Articles

```python
articles_df = pd.read_csv(path+"articles.csv")
```

```python
print(articles_df.shape)
articles_df.head(3)
```

```python
articles_df.isnull().sum()
```

```python
var = 'graphical_appearance_name'
print(len(articles_df[var].unique()))
print(articles_df[var].unique())
```

```python
plt.rcParams['text.color'] = 'black'
plt.rcParams['axes.labelcolor'] = 'black'
plt.rcParams['font.size'] = 10

ax = articles_df['product_group_name'].value_counts(normalize=True).mul(100).round(1).plot(kind = 'bar', color = 'blue', figsize = (10, 5),
                                                    title = 'Bar graph of Product Group Name')
ax.set_xlabel("Product Group Name", fontsize = 12)
ax.set_ylabel("Percentage %", fontsize = 12)
plt.tick_params(labelsize = 10)
plt.grid()
```

From the above graph, it can be seen that the articles of product group 'Garment Upper Body' and 'Garment Lower Body' constitute more than 65 % of the total in the dataset.

```python
ax = articles_df['product_type_name'].value_counts(normalize=True, ascending = False).mul(100).round(1).plot(kind = 'bar', color = 'purple', figsize = (24, 8),
                                                         title = 'Bar graphs on Product Type Name')
ax.set_xlabel('Product Type Name', fontsize = 12)
ax.set_ylabel('Percentage %', fontsize=12)
plt.tick_params(labelsize = 10)
plt.grid()
```

Notice how {Trousers, Dress, Sweater and T-Shirt} dominate the percentage distribution of articles in the articles data. These types of products are sold more than any other product type simply because they are worn and needed by both genders and across all age groups. The product types that follow this subset in percentage distribution, are {Jackets, Shorts, Hoodie}. They are also quite popular in both genders and among people from different age groups but their distribution is comparitively low probably because of the seasonality effect.

```python
ax = articles_df['graphical_appearance_name'].value_counts(normalize=True, ascending = False).mul(100).round(1).plot(kind = 'bar', color = 'green', figsize = (10, 5),
                                                            title = 'Bar graph on Graphical Appearance Name')
ax.set_xlabel('Graphical Appearance Name')
ax.set_ylabel('Frequency')
plt.tick_params(labelsize = 12)
plt.grid()
```

Most of the articles are solid in graphical appearance.

```python
ax = articles_df['section_name'].value_counts(normalize=True, ascending = False).mul(100).round(1).plot(kind = 'bar', color = 'red', figsize = (18, 5),
                                                    title = 'Bar graph of Section Name')
ax.set_xlabel("Section Name", fontsize = 12)
ax.set_ylabel("Frequency", fontsize = 12)
plt.tick_params(labelsize = 10)
plt.grid()
```

Clear observation: Women and Kid section dominates the articles distribution.

```python
ax = articles_df['index_group_name'].value_counts(normalize=True, ascending = False).mul(100).round(1).plot(kind = 'bar', color = 'orange', figsize = (10, 5),
                                                    title = 'Bar graph of Index Group Name')
ax.set_xlabel("Index Group Name", fontsize = 12)
ax.set_ylabel("Frequency", fontsize = 12)
plt.tick_params(labelsize = 10)
plt.grid()
```

Consistent to the distribution of the section column, Ladieswear and Childrenwear dominate the group distritbution of the articles. Articles of Sports group are less than 5%.

```python
ax = articles_df['garment_group_name'].value_counts(normalize=True, ascending = False).mul(100).round(1).plot(kind = 'bar', color = 'brown', figsize = (10, 5),
                                                    title = 'Bar graph of Garment Group Name')
ax.set_xlabel("Garment Group Name", fontsize = 12)
ax.set_ylabel("Frequency", fontsize = 12)
plt.tick_params(labelsize = 10)
plt.grid()
```

By garment group name, Jerseys make about 30% of the articles followed by Accessories which is 10%.

## Customers

```python
customer_df = pd.read_csv(path+"customers.csv")
```

```python
print(customer_df.shape)
customer_df.head(3)
```

```python
customer_df.isnull().sum()
```

```python
str_list = list(customer_df.columns)
str_list.remove('age')
str_list.remove('Active')
for col in str_list:
    customer_df[col].fillna('Null', inplace = True)

customer_df['Active'].fillna(0.0, inplace = True)
customer_df['age'].fillna(0.0, inplace = True)
```

```python
var = 'Active'
print(len(customer_df[var].unique()))
print(customer_df[var].unique())
```

```python
plt.rcParams['font.size'] = 12
plt.rcParams['text.color'] = 'white'
plt.rcParams['axes.labelcolor'] = 'white'
```

```python
labels = customer_df['Active'].value_counts(sort = True).index
active_ = customer_df['Active'].value_counts(sort = True)

colors = ["purple", "maroon"]
explode = (0.1, 0)  # explode 1st slice
 
rcParams['figure.figsize'] = 5, 5
# Plot
plt.pie(active_, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=270,)

plt.title('Distribution of Active Status of Customers',size = 14, color = 'white')
plt.show()
```

Observation: 66% of the customers in the data have inactive status assuming that NULL values in the data only represent inactivity and not missing value. Nevertheless, only 33% of the customers have active status.

```python
plt.rcParams['text.color'] = 'black'
plt.rcParams['axes.labelcolor'] = 'black'
plt.rcParams['font.size'] = 10
```

```python
ax = customer_df['club_member_status'].value_counts(normalize=True).mul(100).round(1).plot(kind = 'bar',
                                                                                        color = 'red', figsize = (8, 5),
                                                                                        title = 'Bar graph of Club Member Status')
ax.set_xlabel("Club Member Status", fontsize = 10)
ax.set_ylabel("Number of Customers", fontsize = 10)
plt.tick_params(labelsize = 10)
plt.grid()
```

80% of the customers have active membership status.
I have yet to figure out how is this consistent with the above graph where "active" customers were just about 33%.

```python
ax = customer_df['age'].hist(bins = 20, color = 'blue', figsize = (10, 5))
ax.set_title('Histogram for Age of Customers')
ax.set_xlabel('Age Bracket', fontsize = 10)
ax.set_ylabel('Number of Customers', fontsize = 10)
plt.xticks(np.arange(min(customer_df['age']), max(customer_df['age'])+1, 5.0))
plt.tick_params(labelsize = 10)
#ax.grid()
```

This histogram shows that most of the customers in the data lie in the age group of 20-30 with the highest proportion in the age-group 20-25. There is also considerate proportion of customers in age group 45-55. Interestingly, we have customers below 5 years of age and above 70 years as well.

## Transactions

```python
trans_df = pd.read_csv(path+"transactions_train.csv")
```

```python
print(trans_df.shape)
trans_df.head(3)
```

```python
trans_df.isnull().sum()
```

```python
plt.rcParams['font.size'] = 12
plt.rcParams['text.color'] = 'white'
plt.rcParams['axes.labelcolor'] = 'white'
```

```python
labels = trans_df['sales_channel_id'].value_counts(sort = True).index
counts_ = trans_df['sales_channel_id'].value_counts(sort = True)

colors = ["blue", "maroon"]
explode = (0.1, 0)  # explode 1st slice
 
rcParams['figure.figsize'] = 5, 5
# Plot
plt.pie(counts_, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=270,)


plt.title('Distribution of Sales Channel ID in Transactions',size = 14, color = 'white')
plt.show()
```

70% of transactions have been carried out through Sales Channel 2.

```python
dates_trans = trans_df['t_dat'].str[:4]
labels = dates_trans.value_counts(sort = True).index
counts_ = dates_trans.value_counts(sort = True)

colors = ['purple', 'maroon', 'green']
explode = (0.1, 0.1, 0.)

rcParams['figure.figsize'] = 5, 5
# Plot
plt.pie(counts_, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=270,)

plt.title('Percentage of transactions in years',size = 14, color = 'white')
plt.show()
```

More than half of the transactions in the dataset took place in the year 2019, followed by 2020 and 2018. Only 13% of the transactions were recorded in 2018.
