```python
import numpy as np 
import pandas as pd 

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        os.path.join(dirname, filename)
        
import seaborn as sns        
import matplotlib.pyplot as plt
```

```python
path = '/kaggle/input/h-and-m-personalized-fashion-recommendations/'
train      = pd.read_csv( path + 'transactions_train.csv')
sample_sub = pd.read_csv(path + 'sample_submission.csv')
articles   = pd.read_csv(path + 'articles.csv')
customers  = pd.read_csv(path + 'customers.csv')
```

```python
train['t_dat'] = pd.to_datetime(train['t_dat'])
print(f"Min time - {train['t_dat'].min()}, Max time - {train['t_dat'].max()}\n")
train['Day'] = train['t_dat'].dt.day
train['Month'] = train['t_dat'].dt.month
train['Year'] = train['t_dat'].dt.year
train = train.drop(['t_dat'], axis ='columns')
```

```python
#get numeric features
numeric_features = [f for f in articles.columns if articles[f].dtype != object]
articles = articles[numeric_features]
articles = articles.drop(['perceived_colour_value_id','perceived_colour_master_id','index_group_no','section_no','garment_group_no','product_type_no','graphical_appearance_no', 'department_no'], axis = 'columns')
articles
```

```python
merged = pd.merge(train, articles, on ="article_id")
```
