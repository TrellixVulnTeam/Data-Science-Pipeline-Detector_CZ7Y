```python
import matplotlib.image as img

def get_image_path(article_num):
    path = '/kaggle/input/h-and-m-personalized-fashion-recommendations/images/'
    article_str = f'{article_num:010d}'
    dir_str = f'{int(article_num/10**7):03d}'
    return path+dir_str+'/'+article_str+'.jpg'

def draw_article_images(articles_seq):
    sns.set_style("white")
    cols = 5
    rows = int(len(articles_seq) / cols)
    fig = plt.figure(figsize=(20, 5*rows))
    for i, idx in enumerate(articles_seq):
        try:
            article_num = idx
            path = get_image_path(article_num)
            image = img.imread(path)
            ax = fig.add_subplot(rows, cols, i+1)
            ax.imshow(image)
        except FileNotFoundError:
            print(path+' : file not found')
    plt.show()
```

### Images of top selling article by total sales

```python
# drawing images of top selling articles by total sales
draw_article_images(top_selling_article_by_total_sales.index)
```

# Price Outliers By Category

```python
# merge transaction + customer + article
transactions_to_be_merged = transactions_train[['t_dat', 'customer_id', 'article_id', 'price']]
article_to_be_merged = articles[['article_id', 'prod_name', 'product_type_name', 'product_group_name', 'index_group_name']]
customers_to_be_merged = customers[['customer_id', 'age']]

transactions_merged = transactions_to_be_merged.merge(article_to_be_merged, on='article_id', how='left')
transactions_merged = transactions_merged.merge(customers_to_be_merged, on='customer_id', how='left')

transactions_merged.head()
```

```python
produt_group_names = transactions_merged['product_group_name'].unique()
produt_group_names
```

```python
box_color = ['xkcd:pink', 'xkcd:blue', 'xkcd:teal', 'xkcd:goldenrod']
sns.set_style("darkgrid")
for i, name in enumerate(produt_group_names):
    figure, ax = plt.subplots(figsize=(20,1))
    ax = sns.boxplot(data=transactions_merged[transactions_merged['product_group_name']== name], x='price', color=box_color[i%len(box_color)])
    plt.xticks(np.arange(-0.05, 0.7, step=0.05))
    ax.set_ylabel(name)
    plt.show()
ax.set_xlabel('Price outliers')
plt.show()
```

```python
# how much top selling n% articles contribute to overall sales
n_pct_sales = [article_sorted_by_sales.head(int(article_sorted_by_sales.shape[0]*0.01*(n))).sum() / total_sales * 100 for n in range (1, 100+1)]
# n_pct_sales
```

```python
# top selling n% articles => contribution percentage to total sales
for n in range(100):
    print("{:3}% => {:3.2f} % sales (+{:3.2f})%".format(n+1, n_pct_sales[n], n_pct_sales[n] - n_pct_sales[n-1] if n>0 else 0))
    print("-------------------------------") if (n+1) % 10 == 0 and (n+1) != 0 else 0
```

```python
sns.set_style("darkgrid")
figure, ax = plt.subplots(figsize=(6,6))
ax = sns.lineplot(x=range(1, 100+1), y=n_pct_sales, color='xkcd:blue')
ax.set_xlabel('n percent articles')
ax.set_ylabel('sales')
plt.show()
```

### Top 20% account for 80% of total sales  => Pareto principle
