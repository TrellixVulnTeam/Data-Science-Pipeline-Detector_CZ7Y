```python
import numpy as np
import pandas as pd
# import pickle
```

```python
def get_features(dataframe, unimportant=set()):
    feature_columns = dataframe.columns
    columns_to_remove = []
    for feature in feature_columns:
        if not feature.startswith('feature_'):
            columns_to_remove.append(feature)
    feature_columns = feature_columns.drop(columns_to_remove)
    return sorted(feature_columns)

def get_X_array(df, feature_columns):
    return df.loc[:, feature_columns].values
```

```python
train = pd.read_csv('/kaggle/input/jane-street-market-prediction/train.csv', nrows=3)

feature_columns = get_features(train)

feature_na_dict = {}
for cc in feature_columns:
    feature_na_dict[cc] = -999
    
na_arr = np.array([feature_na_dict[cc] for cc in feature_columns])

#mdl_0 = pickle.load(open('../input/pretrained/jane_model_1.pickle', 'rb'))
```

```python
import janestreet
env = janestreet.make_env()
iter_test = env.iter_test()

for (test_df, sample_prediction_df) in iter_test:
    
    xar = get_X_array(test_df, feature_columns)
    xar = xar - na_arr
    xar = np.nan_to_num(xar, nan=0)
    xar = xar + na_arr
#     p = mdl_0.predict(xar)[0]
    p = 0.5
    sample_prediction_df.action = 1 if p > 0 else 0
    env.predict(sample_prediction_df)
```
