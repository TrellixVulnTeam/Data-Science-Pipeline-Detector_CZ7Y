**Check for missing values**

```python
def check_missings(dataset):
    
    df_NA = dataset.isnull().sum()
    df_NA = df_NA.drop(df_NA[df_NA == 0].index).sort_values(ascending = False)
    NA_perc = (dataset.isnull().sum()/dataset.isnull().count()*100).sort_values(ascending = False)
    NA_perc = NA_perc.drop(NA_perc[NA_perc == 0].index)

    return pd.concat([df_NA, NA_perc], axis=1, keys=['Total', 'Percent'])
```

```python
check_missings(train)
```
