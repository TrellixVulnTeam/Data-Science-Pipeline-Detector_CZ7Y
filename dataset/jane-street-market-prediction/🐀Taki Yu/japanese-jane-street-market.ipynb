# Jane Street Market Predictionコンペについて
## コンペ主催者Jane Streetとは

- アメリカでは実力を持っている完全自己資本(企業の総資本のうち返済する必要がない資本)の投資会社
- 日本にも千葉に支社があるようです。
- 事業内容は、証券トレーディング
- [公式サイト](http://www.janestcapital.com)
- [日本の企業情報](https://www.daijob.com/jobs/company/2537)

## コンペの目的
- ジェーンストリート市場予測
- 最初の３ヶ月間(notebook提出期限 2021/02/22 )主要な証券取引所からの市場データを使用してリターンを最大化するモデルを作る
- 2021/08/23までモデルの予測性を実際の市場、リーダーボードでフィードバックを受け取ります。

## メモ
- 利益の取引は、動きが早く複雑な要因が多々あるため、予測が難しい
- 災害など予測できないことが多々起きそうだ(筆者の主観)
- 時系列なのでまずは、図にして可視化したほうが良さそうだ


# 扱うデータ

- train.csv(訓練データ。履歴データとリターンが含まれる)
  - ts_id
    - ts_idは時間の順序。
  - date
    - 取引の日を表す整数
  - weight
    - 重み。重み＝０は、スコアの評価に関係なし。完全性のために意図的に入れた
  - resp
    - resp値(後ほど、詳しく調べる)
  - resp_{1,2,3,4} 
  - feature_{0...129}
- example_test.csv 
  - テストデータのサンプル。実際のテストデータは、時系列APIを使ってデータがセットされる
  - データ構造は、train.csvの構造に近いが以下の点が異なる
    - リターン(resp_)が設定されていない(そりゃそうだ)
- example_sample_submission.csv
 - 提出形式のファイル
 - ts_id
 - action
   - 0・・・pass(何もしない)
   - 1・・・取引
- features.csv
  - featuresに対する特徴のようなもの (tag_0〜tag_28)
