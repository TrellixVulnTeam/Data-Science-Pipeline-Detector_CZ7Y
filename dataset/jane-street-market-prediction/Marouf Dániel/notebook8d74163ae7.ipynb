```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import janestreet

from lightgbm import LGBMClassifier    
```

```python
train = pd.read_csv("/kaggle/input/jane-street-market-prediction/train.csv")
```

```python
train["action"] = train["resp"] > 0
```

```python
train = train[train["date"] > 85]
```

```python
features = [col for col in train.columns if "feature" in col]
```

```python
model = LGBMClassifier()
model.fit(train[features], train["action"])
```

```python
env = janestreet.make_env() # initialize the environment
iter_test = env.iter_test() # an iterator which loops over the test set
for (test_df, sample_prediction_df) in iter_test:
    sample_prediction_df["action"] = model.predict(test_df[features]).astype(int)
    env.predict(sample_prediction_df)
```
