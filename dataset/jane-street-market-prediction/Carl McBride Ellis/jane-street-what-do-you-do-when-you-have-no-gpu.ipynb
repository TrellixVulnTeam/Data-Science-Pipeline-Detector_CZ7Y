```python
import numpy  as np
import pandas as pd
import janestreet
env = janestreet.make_env()
iter_test = env.iter_test()
for (test_df, sample_prediction_df) in iter_test:
    feature_35 = test_df['feature_35'].item()
    feature_43 = test_df['feature_43'].item()
    sample_prediction_df.action = 1 if (feature_35 > 0) and (feature_43 > 0) else 0
    env.predict(sample_prediction_df)
```

*Notes*: 
* `feature_35` is one of the [17 'resp' related features](https://www.kaggle.com/carlmcbrideellis/jane-street-eda-of-day-0-and-feature-importance), and `feature_43` is the third of the three Tag 14 features, having the most 'biased' lag plot)
