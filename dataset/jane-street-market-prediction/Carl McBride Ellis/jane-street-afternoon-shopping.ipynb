```python
import numpy  as np
import pandas as pd
import janestreet
env = janestreet.make_env()
iter_test = env.iter_test()
for (test_df, sample_prediction_df) in iter_test:
    
    feature_0   = test_df['feature_0'].item()  # 1 == buy, -1 == sell
    feature_64  = test_df['feature_64'].item() # the 'clock' feature
    
    # 'selling' features
    feature_43  = test_df['feature_43'].item()
    feature_117 = test_df['feature_117'].item()
    
    # 'buying' features
    feature_8   = test_df['feature_8'].item()
    feature_9   = test_df['feature_9'].item()
    
    # selling
    if feature_0 < 0:
        sample_prediction_df.action = 1 if (feature_117 > 0.2) and (feature_43 > 0) else 0
        
    # buying (...but only after a nice lunch)
    else:
        sample_prediction_df.action = 1 if (feature_64  > 0.7) and (feature_8 > 0.1) and (feature_9 > 0.2) else 0
        
    env.predict(sample_prediction_df)
```

**Note:** `0.7` is the lower limit of the lunchtime discontinuity in `feature_64`.
