# Volatility Correlations

I'm interested in identifying potential volitility features, so here I'm going to look at the correlation with abs(resp). I'm also throwing in resp, weight and (resp * weight) too for good measure.

```python
%%time
corrs = pd.DataFrame({'resp': train.corrwith(train['resp']),
                      'weight': train.corrwith(train['weight']),
                      'resp*weight': train.corrwith(train['resp'] * train['weight']),
                      'resp_abs': train.corrwith(train['resp'].abs())})
```

```python
corrs.style.background_gradient(cmap='coolwarm', axis=None)
```

There are a few features here which are correlated, perhaps most interesting is the block of features 41-43. Feature 43 shows the highest correlation at ~0.41, with 42 at ~0.35 and 41 at ~-0.16. All three of these features only have tag_14 set to true in features.csv, and these are the only features with this tag enabled. It is tempting to consider the idea that these features may be different volatility measures (possibly over different time horizons), although the negative correlation of abs(resp) with feature 41 suggests this is not the case.
