```python
import os
import gc
import sys
import time
import tqdm
import random
import numpy as np
import pandas as pd
import seaborn as sns
import datatable as dt
import matplotlib.pyplot as plt
plt.style.use('bmh')
```

```python
folder_path = '../input/jane-street-market-prediction/'
train = dt.fread(folder_path + 'train.csv').to_pandas()
features = dt.fread(folder_path + 'features.csv').to_pandas()
```

```python
def plot_correlation_over_time(groups):
    # plotting correlation between first 3 features over different time horizons
    groups_len = len(np.arange(1, 500, groups))
    fig, axes = plt.subplots(groups_len//2, 2, figsize=(20, (groups_len//2)*8))

    # loop over first day in selected group spans
    for i, ax in zip(np.arange(1, 500, groups), axes.flatten()):
        group_days = [i for i in range(i, i+groups)]
        group = train.loc[train.date.isin(group_days)]

        # calculate pearson correlation
        corr_mat = group.corr(method='pearson')

        sns.heatmap(corr_mat, ax=ax, cmap='coolwarm');
        ax.set_title(f'Days between {group_days[0]} and {group_days[-1]}')
```

```python
plot_correlation_over_time(6);
```

```python
plot_correlation_over_time(10);
```

```python
plot_correlation_over_time(30);
```
