## Convert training data from CSV to Parquet to dramatically speed-up reading

Dataset: https://www.kaggle.com/gdonchyts/jane-street-market-prediction-parquet

Memory: 5.8GB -> 1.23GB

Reading speed: 1min 55sec -> 5 sec

```python
import pandas as pd
import numpy as np
```

```python
%%time
train = pd.read_csv('../input/jane-street-market-prediction/train.csv')
```

```python
train.info()
```

```python
#limit memory use
train = train.astype({c: np.float32 for c in train.select_dtypes(include='float64').columns}) 
```

```python
train.to_parquet('./train.parquet')
```

```python
%%time
train = pd.read_parquet('./train.parquet')
```
