```python
from pprint import pprint
import os
import sys
import numpy as np
import pandas as pd
from time import time, sleep
import json
from datetime import datetime
import warnings
from collections import defaultdict, Counter
import plotly.express as px
import plotly.graph_objects as go
```

```python
!wget 'https://www.kaggle.com/c/jane-street-market-prediction/leaderboard.json?includeBeforeUser=true&includeAfterUser=false' -O leaderboard.json
```

```python
with open("leaderboard.json") as f:
    jsn = json.load(f)
```

```python
scores = []
tiers = []
medals = []
entries = []
teams = []
for user in jsn['beforeUser']+jsn['afterUser']:
    for teammember in user['teamMembers']:
        tiers.append(teammember['tier'])
    scores.append(user['score'])
    medals.append(user['medal'])
    entries.append(user['entries'])
    teams.append(user['teamName'])
```

## Top 15 Teams

```python
for i in range(100):
    scores[i] = float(scores[i])
score_df = pd.DataFrame({"Team":teams[0:100],"Public Score":scores[0:100]})
```

```python
best_ten = score_df.iloc[0:15]
fig = px.bar(best_ten.iloc[::-1], x='Public Score', y='Team',
              color='Public Score',height=700, title='Top 15 Teams (With Public Score)',text='Public Score')
fig.show()
```

## Top 100 Teams

```python
fig = px.bar(score_df.iloc[::-1], x='Public Score', y='Team',
              color='Public Score',height=4000, title='Top 100 Teams (With Public Score)',text='Public Score')
fig.show()
```

```python
ranking = 1
for i,j in zip(teams,scores):
    print("Rank: " + str(ranking) + " Team:" + i + "     Score: " + str(j))
    ranking += 1
    if ranking >= 101:
        break
```

## Teams in Medal Zone

### Gold Medal

```python
rank = 1
rank_list = []
gold_teams = []
for i,j in zip(medals,teams):
    if i == 'gold':
        rank_list.append(rank)
        gold_teams.append(j)
        rank += 1
```

```python
fig = go.Figure(data=[go.Table(
    header=dict(values=['Rank','Team'],
                fill_color='paleturquoise',
                line_color='blue',
                align='left'),
    cells=dict(values=[rank_list,gold_teams],
               fill_color='lavender',
               align='left'))
])
fig.update_layout(title='Gold Medal Zone')
fig.show()
```

### Silver Medal

```python
silver_teams = []
rank_silver = []
for i,j in zip(medals,teams):
    if i == 'silver':
        rank_silver.append(rank)
        silver_teams.append(j)
        rank += 1
```

```python
fig = go.Figure(data=[go.Table(
    header=dict(values=['Rank','Team'],
                fill_color='paleturquoise',
                line_color='blue',
                align='left'),
    cells=dict(values=[rank_silver,silver_teams],
               fill_color='lavender',
               align='left'))
])
fig.update_layout(title='Silver Medal Zone (You can scroll the table)')
fig.show()
```

### Bronze Medal

```python
bronze_teams = []
rank_bronze = []
for i,j in zip(medals,teams):
    if i == 'bronze':
        rank_bronze.append(rank)
        bronze_teams.append(j)
        rank += 1
```

```python
fig = go.Figure(data=[go.Table(
    header=dict(values=['Rank','Team'],
                fill_color='paleturquoise',
                line_color='blue',
                align='left'),
    cells=dict(values=[rank_bronze,bronze_teams],
               fill_color='lavender',
               align='left'))
])
fig.update_layout(title='Bronze Medal Zone (You can scroll the table)')
fig.show()
```

## Public Score in the Medal Zone

```python
score_hist = [0,0,0,0,0]
for i in scores[0:423]:
    
    if float(i) >= 11000:
        score_hist[0] += 1
    elif float(i) >= 10000:
        score_hist[1] += 1
    elif float(i) >= 9000:
        score_hist[2] += 1
    elif float(i) >= 8000:
        score_hist[3] += 1
    else:
        score_hist[4] += 1
```

```python
team_score = pd.DataFrame(data=score_hist,index=['11000.000-','10000.000 - 11000.000','9000.000 - 10000.000'
                                                ,'8000.000 - 9000.000','- 8000.000'],columns=['Number of Teams'])
```

```python
fig = px.bar(team_score, x='Number of Teams', y=team_score.index,labels={'index':'Public Score'})
fig.show()
```

## Entries

```python
ranks = []
for i in range(423):
    ranks.append(i+1)
fig = go.Figure(data=[go.Bar(
            x=ranks, y=entries
        )])
fig.update_layout(title_text='Entries in the Medal Zone',xaxis_title="Rank",yaxis_title='Entries')
fig.show()
```
