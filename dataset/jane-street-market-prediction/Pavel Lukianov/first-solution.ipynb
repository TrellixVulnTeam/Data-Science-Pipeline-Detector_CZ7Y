```python
train = pd.read_csv('/kaggle/input/jane-street-market-prediction/train.csv')
```

```python
features = [col for col in list(train.columns) if 'feature' in col]
train = train[train['weight'] != 0]
train['action'] = (train['resp'].values > 0).astype(int)
train = train.fillna(-999)
X_train = train.loc[:, features]
y_train = train.loc[:, 'action']
```

```python
params = {'n_estimators': 997,
 'max_depth': 3,
 'learning_rate': 0.05225334248788404,
 'subsample': 0.6628646821072358,
 'colsample_bytree': 0.7378948607289495,
 'gamma': 8,
 'missing': -999,
 'tree_method': 'gpu_hist'}
```

```python
clf0 = xgb.XGBClassifier(n_estimators = 997, max_depth = 3, learning_rate = 0.05225334248788404, subsample = 0.6628646821072358,
colsample_bytree = 0.7378948607289495, gamma = 8, missing = -999, tree_method = 'gpu_hist')
```

```python
clf0
```

```python
scaler = StandardScaler()
```

```python
clf = Pipeline(steps=[('scaler', scaler), ('xgb', clf0)])
```

```python
clf.fit(X_train, y_train)
```

```python
for (test_df, prediction_df) in env.iter_test():
    X_test = test_df.loc[:, features]
    y_preds = clf.predict(X_test)
    prediction_df.action = y_preds
    env.predict(prediction_df)
```
