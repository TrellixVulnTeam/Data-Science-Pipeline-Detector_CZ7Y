# Loading CSV files is much slower than many other choices.

This is a simple file to save the JaneStreet data to a feather. If you import this file into yours using the "+ Add data" button in the upper right of your screen, then you can load it using this line:  train = pd.read_feather('/kaggle/input/janestreet-data/train.feather')

This is not a new idea. I'm sharing it, in part, so that I can use it in other notebooks I want to post. Even so, it's a nice light-weight file that might save you some time.

If you want to know much more, check out:  https://www.kaggle.com/rohanrao/tutorial-on-reading-large-datasets

```python
import numpy as np
import pandas as pd
import os
```

```python
DATA_PATH = '../input/jane-street-market-prediction'
```

## Main

First step is to import the data.

```python
# Main

# Import the data from train.csv.
file = 'train.csv'
path = os.path.join(DATA_PATH, file)
X_train = pd.read_csv(path)
print(f'file: {file}  shape: {X_train.shape}')

# Save the data in feather format.
file = 'train.feather'
path = os.path.join(os.getcwd(), file)
X_train.to_feather(os.path.join(os.getcwd(), 'train.feather'), version=1)
print('Data Saved')

X_train.head()
```
