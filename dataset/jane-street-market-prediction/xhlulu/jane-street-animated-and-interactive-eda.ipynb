```python
import plotly.express as px
import pandas as pd
```

```python
train = pd.read_csv('/kaggle/input/jane-street-market-prediction/train.csv')
```

## Histogram over all features at a fixed time

```python
date = 0
n_features = 130

features = [f'feature_{i}' for i in range(1, n_features)]
fig = px.histogram(train[train.date == date], x=features, animation_frame='variable', range_y=[0, 600], range_x=[-7, 7])
fig.show()
```

## Histogram of returns over time

```python
t_start = 0
t_end = 30

fig = px.histogram(
    train[(train.date >= t_start) & (train.date <= t_end)], 
    x=['resp_1', 'resp_2', 'resp_3', 'resp_4'], 
    facet_col='variable', animation_frame='date')
fig.show()
```

## Colored scatter plot of return over time

```python
# One feature on x-axis, 2nd feature as color
x, cl = 1, 2
t_start, t_end = 0, 5
resp = 'resp_1'

fig = px.scatter(
    train[(train.date >= t_start) & (train.date <= t_end)], 
    x=f'feature_{x}', 
    y=resp, 
    color=f'feature_{cl}',
    animation_frame='date'
)
fig.show()
```

## Scatter matrix for a fixed `resp` and time

```python
feature_ix = [1,10,20,30,40]
resp = 'resp_1'
date = 0

fig = px.scatter_matrix(
    train[train.date == date], 
    dimensions=[f'feature_{i}' for i in feature_ix],
    color=resp
)
fig.show()
```
