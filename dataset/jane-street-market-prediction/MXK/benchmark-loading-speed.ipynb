# Libs

```python
import csv
import numpy as np
from numpy import genfromtxt
from numba import njit
import cudf
import cupy
import pandas as pd
import datatable as dt
import pickle
import joblib
import feather
#import plotly.express as px
import matplotlib.pyplot as plt

data_path = '/kaggle/input/jane-street-market-prediction/train.csv'
# saved_data_path = '../input/name of the netebook/name of the data file' if you import the data from another notebook
saved_data_path = '../input/benchmark-data-loading-speed/'
```

# Load saved data

```python
#################################### 1- Pythonâ€™s Built-in CSV parser ####################################
# Result : Your notebook tried to allocate more memory than is available. It has restarted.
```

```python
#################################### 2- Numpy ####################################
speed_np = %timeit -o np.load('../input/benchmark-loading-speed/data.npy')
```

```python
#################################### 4- Pandas ####################################
speed_pd =  %timeit -o  pd.read_csv(data_path)
```

```python
#################################### 5- Datatable ####################################
speed_dt =  %timeit -o  dt.fread(data_path)
```

```python
#################################### 6- cuDF ####################################
speed_cu =  %timeit -o  data_cudf = cudf.read_csv(data_path)
```

```python
#################################### 7- cuPY ####################################
#data = cupy.load('/./data_dt.pickle', allow_pickle=False) # if the loaded file is pickled then allow_pickle=True
# Result : Only load pickled files
```

```python
#################################### 8- Pickle + Datatable ####################################
speed_pickle =  %timeit -o  pickle.load(open(saved_data_path + 'data_dt.pickle', 'rb'))
```

```python
#################################### 9- Joblib + Datatable ####################################
speed_joblib =  %timeit -o  joblib.load(saved_data_path + '/data_dt.joblib')
```

```python
#################################### 10- Feather + Pandas ####################################
speed_feather =  %timeit -o  pd.read_feather(saved_data_path + './data_pd.feather')
```

```python
#################################### 11- Parquet + Pandas ####################################
speed_parquet =  %timeit -o  pd.read_parquet(saved_data_path + './data.parquet')
```

```python
#################################### 12- Jay + Datatable ####################################
speed_jay =  %timeit -o  dt.fread(saved_data_path + './data_dt.jay')
```

```python
#################################### 13- hdf5 + Pandas ####################################
speed_hdf5 =  %timeit -o pd.read_hdf(saved_data_path + './data.h5', "data")
```

### Benchmark

```python
speed = ["{:.2f}".format(speed_np.average) ,"{:.2f}".format(speed_pd.average), "{:.2f}".format(speed_dt.average), "{:.2f}".format(speed_cu.average), "{:.2f}".format(speed_pickle.average), "{:.2f}".format(speed_joblib.average), "{:.2f}".format(speed_feather.average), "{:.2f}".format(speed_parquet.average), "{:.4f}".format(speed_jay.average), "{:.2f}".format(speed_hdf5.average)]
speed_name = ['Numpy', 'Pandas', 'Datatable', 'cuDF', 'Pickle', 'Joblib', 'Feather', 'Parquet', 'Jay', 'hdf5']

for row in range(len(speed)):
    print(speed_name[row] + ' = ', speed[row], 's')
    
speedy = pd.DataFrame(index=speed_name)
speedy["Lib name"] = speed_name
speedy["Lib speed"] = list(map(float, speed))
speedy.sort_values(by=['Lib speed'], ascending=True, inplace=True)
```

```python
speedy.to_pickle("./speedy.pkl")
# pd.read_pickle("./dummy.pkl")
```

```python
# plotly express
#fig = px.bar(speedy, y='Lib speed', x='Lib name', text='Lib speed', title='Benchmarking Data loading speed in seconds')
#fig.show() # the fig is not showing

# pandas vis
#speedy_ax = speedy.plot.bar(rot=90) # bad viz, yet simple


# pyplot
plt.rcdefaults()
fig, ax = plt.subplots()

y_pos = np.arange(len(speedy["Lib name"]))
performance = 3 + 10 * np.random.rand(len(speedy["Lib name"]))
error = np.random.rand(len(speedy["Lib name"]))

ax.barh(y_pos, speedy["Lib speed"], xerr=error, align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(speedy["Lib name"])
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Speed in Seconds')
ax.set_title('Benchmarking Data loading speed in seconds')

plt.show()
```
