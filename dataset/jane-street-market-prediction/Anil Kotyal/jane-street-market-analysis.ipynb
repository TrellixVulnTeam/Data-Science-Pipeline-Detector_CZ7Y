```python
#from sklearn.impute import SimpleImputer
#imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
#imputer=imputer.fit_transform(Data)

train_df=Data.fillna(Data.mean())
example_test.fillna(example_test.mean(),inplace=True)
```

```python
train_df.head(5)
```

```python
train_df.isna().sum().sum()
```

```python
train_df['action']=np.where(train_df[['resp_1','resp_2','resp_3','resp_4','resp']].mean(axis=1)>0,1,0)
```

```python
#train_df['action'] = np.where((train_df.resp_1 > 0) & (train_df.resp_2 > 0) & (train_df.resp_3 > 0) & (train_df.resp_4 > 0) & (train_df.resp > 0),1,0)
```

```python
train_df.head()
```

```python
plt.hist(train_df['action'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
#Train test-split
X_train, X_test, y_train, y_test = train_test_split(train_df.drop(columns = ['date', 'weight', 'resp_1', 'resp_2', 'resp_3', 'resp_4','resp', 'ts_id','action']),train_df['action'],test_size=0.25)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# Random Forest classification

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RFC=RandomForestClassifier(n_estimators=10,criterion="entropy",random_state=0)
RFC.fit(X_train,y_train)
```

```python
y_predict=RFC.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_predict)
accuracy=(cm[0,0]+cm[1,1])/np.sum(cm)
print(accuracy)
```

```python
y_predict=RFC.predict(X_test)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_predict)
accuracy=(cm[0,0]+cm[1,1])/np.sum(cm)
print(accuracy)
```

# checking accuracy for example_test data

```python
example_test.head()
```

```python
df_submission= RFC.predict(example_test.drop(columns=['ts_id','date','weight']))
```

```python
df_submission=pd.DataFrame(df_submission,columns=['y_pred'])
df_submission['ts_id']=example_test['ts_id']
```

```python
example_samp_sub=example_samp_sub.set_index('ts_id')
df_submission=df_submission.set_index('ts_id')
print(example_samp_sub.head())
print(df_submission.head())
```

```python
Final_permis=pd.merge(example_samp_sub,df_submission,how='inner',left_index=True,right_index=True)
Final_permis.head()
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Final_permis.action,Final_permis.y_pred)
accuracy=(cm[0,0]+cm[1,1])/np.sum(cm)
print(accuracy)
```
