```python
import numpy as np
import pandas as pd
import datatable as dt
```

```python
path_train_data = '/kaggle/input/jane-street-market-prediction/train.csv'
```

```python
%%time

dt_train = dt.fread(path_train_data)
df_train = dt_train.to_pandas()
df_train.info(max_cols=999)
del dt_train
```

### Maximization metric on train data

```python
df_4metric = df_train[['date', 'weight', 'resp']].copy()
print(df_4metric.shape)
df_4metric.head()
```

```python
df_4metric['max_p'] = df_4metric['weight'] * df_4metric['resp'] * (df_4metric['resp'] > 0)
df_4metric.head(10)
```

```python
df_4metric_days = df_4metric[['date', 'max_p']].groupby('date').sum().reset_index()
print(df_4metric_days.shape)
df_4metric_days.head(10)
```

```python
p = df_4metric_days['max_p'].values

t = (np.sum(p)/(np.sqrt(np.sum(p**2))))*np.sqrt(250/len(p))

u = min(max(t, 0), 6) * np.sum(p)

print(f"Maximum score is: {u:.3f}")
```

```python
p1 = p - 0.01
t1 = (np.sum(p1)/(np.sqrt(np.sum(p1**2))))*np.sqrt(250/len(p1))
u1 = min(max(t1, 0), 6) * np.sum(p1)
print(f"Maximum score is: {u1:.3f}")
```
