# Jane Street: API explain

This is a simple exploration of the [evaluation code](https://www.kaggle.com/c/jane-street-market-prediction/overview/evaluation) and the janestreet module. It is nice to have an overview of the API if you do things localy first.
Let us import the module and initialize the environment:

````python
import janestreet
env = janestreet.make_env() # initialize the environment
iter_test = env.iter_test() # an iterator which loops over the test set
````

The `iter_test` is an iterator, i.e. a generator, so in order to have a single value we use the `next` statement:

````python
a = next(iter_test)
````

If you run more than once in a single session the API returns:
```
You must call `predict()` successfully before you can continue with `iter_test()`
```
The conclusion is that if we want to progress with reading the test values the full cycle must be performed. The yield of the test generator returns a tuple. So lets move to the next value properly.

````python
a[1].action = 1
env.predict(a[1])
b = next(iter_test)
````

````python
print(f"type(a)={type(a)}")
print(f"a=\n{a}")
print(f"b=\n{b}")
````

So we have for the two elements of the tuple:

````python
print(f"type(a[0])={type(a[0])}")
print("a[0]=")
a[0]
````

````python
print(f"type(a[1])={type(a[1])}")
print("a[1]=")
a[1]
````

Speciffic elements can be accessed as:

````python
a[0].loc[0, "feature_3"]
````

And for the next element the row index is increasing

````python
b[0].loc[1, "feature_3"]
````
