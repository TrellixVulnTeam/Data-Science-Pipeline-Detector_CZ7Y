```python
import pandas as pd
jane = pd.read_csv('../input/jane-street-market-prediction/train.csv')
jane.head()
```

```python
jane.columns.tolist()
```

```python
jane['action']=(jane['resp'].values>0).astype(int)
```

```python
corr = jane.corr()
corr.style.background_gradient(cmap='coolwarm')

```

```python
#Showing features with correlation with action more than 0.02
cor_target = abs(corr["action"])#Selecting highly correlated features
relevant_features = cor_target[cor_target>0.02]
relevant_features
```

```python
#Selecting highly correlated features with resp
cor_target = abs(corr["resp"])
relevant_features = cor_target[cor_target>0.025]
relevant_features
```

```python
#highly correlated features with "action"
corr_values = corr.abs()
highly_correlated = corr_values.where(np.triu(np.ones(corr_values.shape), k=1).astype(np.bool))

# Find features with correlation greater than 0.9
correlated_columns = [column for column in highly_correlated.columns if any(highly_correlated[column] > 0.9)]

correlated_columns
```

```python
correlated_columns.pop(0)
correlated_columns.pop()
correlated_columns
```

```python
#new dataset
jane = jane.drop(correlated_columns, axis=1)

# Show new dataframe
jane.head()
```
