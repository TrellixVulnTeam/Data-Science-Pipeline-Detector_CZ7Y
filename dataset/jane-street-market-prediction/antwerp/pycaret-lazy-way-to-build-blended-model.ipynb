```python
from pycaret.regression import *
```

```python
df_train = pd.read_csv("/kaggle/input/jane-street-market-prediction/train.csv")
```

```python
df_train_ = df_train.drop(["resp_1", "resp_2", "resp_3", "resp_4"], axis=1)
```

```python
df_train_ = df_train_.sample(n=55000)
```

```python
exp = setup(df_train_, target="resp", use_gpu=True, silent=True)
```

```python
compare_models()
```

```python
omp = create_model("omp")
br = create_model("br")
lar = create_model("lar")
lightgbm = create_model("lightgbm")
xgboost = create_model("xgboost")
blend = blend_models(estimator_list = [omp, br, lar, lightgbm, xgboost])
```

```python
tuned = tune_model(blend, search_library = "optuna")
```

```python
evaluate_model(tuned)
```

```python
df_test = pd.read_csv("/kaggle/input/jane-street-market-prediction/example_test.csv")
```

```python
result = predict_model(tuned, data = df_test)
```

```python
result[["ts_id", "Label"]].to_csv("submission.csv", index=False)
```
