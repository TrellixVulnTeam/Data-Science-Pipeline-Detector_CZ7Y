1. Import dataset

# 2. Copy & rename packages

Notice:
Due to the Kaggle Dataset's design, `*tar.gz` will be uncompressed when creating dataset.
I didn't want the package to be uncompressed (For easier installation). 

To prevent being uncompressed, I renamed `*.tar.gz` into `*tar.gz.tmp` before creating dataset. 

Thus you need to rename `*tar.gz.tmp` into `*tar.gz` back before `pip install`.

```python
!cp -r /kaggle/input/mlflow1131 /kaggle/working
!mv /kaggle/working/mlflow1131/alembic-1.4.1.tar.gz.tmp /kaggle/working/mlflow1131/alembic-1.4.1.tar.gz
!mv /kaggle/working/mlflow1131/databricks-cli-0.14.1.tar.gz.tmp /kaggle/working/mlflow1131/databricks-cli-0.14.1.tar.gz
!mv /kaggle/working/mlflow1131/prometheus_flask_exporter-0.18.1.tar.gz.tmp /kaggle/working/mlflow1131/prometheus_flask_exporter-0.18.1.tar.gz
!ls /kaggle/working/mlflow1131
```

We're ready to install!

# 3. Install

```python
!pip install /kaggle/working/mlflow1131/*
```
