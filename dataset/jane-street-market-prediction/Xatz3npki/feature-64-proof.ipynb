```python
import numpy as np 
import pandas as pd 
import plotly.express as px
```

```python
train = pd.read_csv('../input/jane-street-market-prediction/train.csv')
train = train.iloc[:100000, :]
train.fillna(train.mean(),inplace=True)
train = train.astype({c: np.float32 for c in train.select_dtypes(include='float64').columns}) 

ll = [7, 11, 13, 15, 9, 8, 12, 14, 16, 10, 17, 21, 23, 25, 19, 18, 22, 24, 26, 20, 27, 31, 33, 35, 29, 28, 32, 34, 36, 30, 55, 57, 58, 59, 56, 72, 74, 75, 76, 73, 78, 80, 81, 82, 79, 84, 86, 87, 88, 85, 90, 92, 93, 94, 91, 96, 98, 99, 100, 97, 102, 104, 105, 106, 103, 108, 110, 111, 112, 109, 114, 116, 117, 118, 115, 120, 124, 126, 128, 122, 121, 125, 127, 129, 123]
features_sorted = ['feature_' + str(i) for i in ll ]

X = train[features_sorted]
Xcs = X.cumsum(axis=0)
Xcs['date'] = train['date']
Xcs['feature_64'] = train['feature_64']

Xcs
```
