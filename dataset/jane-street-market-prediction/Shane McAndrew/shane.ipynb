```python
import datatable as dt
X = dt.fread('/kaggle/input/jane-street-market-prediction/train.csv').to_pandas()
```

```python
X = X[X.date > 85]
y = np.where((X.resp > 0), 1, 0)
X.drop(['resp_1', 'resp_2', 'resp_3', 'resp_4', 'resp', 'ts_id'], axis=1, inplace=True)
```

```python
from sklearn.experimental import enable_hist_gradient_boosting
from sklearn.ensemble import HistGradientBoostingClassifier
clf = HistGradientBoostingClassifier(random_state=42).fit(X, y)
```

# test only
del X, y

```python
import janestreet
env = janestreet.make_env() # initialize the environment
iter_test = env.iter_test() # an iterator which loops over the test set
```
