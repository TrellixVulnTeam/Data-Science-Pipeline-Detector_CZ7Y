```python
import janestreet
env = janestreet.make_env() # initialize the environment
iter_test = env.iter_test() # an iterator which loops over the test set

for (test_df, sample_prediction_df) in iter_test:
    sample_prediction_df.action = kool_model.predict(test_df) #make your 0/1 prediction here
    env.predict(sample_prediction_df) 
```

```python
sample_prediction_df = sample_prediction_df.rename_axis('ts_id').reset_index()
sample_prediction_df.to_csv('submission.csv', index=False)
```
