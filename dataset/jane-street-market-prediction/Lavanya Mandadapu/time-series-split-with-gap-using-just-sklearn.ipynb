## Simple Example

Create a simple toy dataset.

```python
from sklearn.model_selection import TimeSeriesSplit

X = [i for i in range(100)]
y = [ascii(i) for i in X]
print("Full X:", X)
print("Full y", y)
```

Usage of TimeSeriesSplit without gaps.

```python
splitter = TimeSeriesSplit(n_splits=8)
x_y_indices = splitter.split(X, y)
for train_i, test_i in x_y_indices:
    print("train", train_i, "\ttest_i", test_i, "\n")
```

Usage of TimeSeriesSplit with gap.

```python
splitter = TimeSeriesSplit(n_splits=8, gap=1)
x_y_indices = splitter.split(X, y)
for train_i, test_i in x_y_indices:
    print("train", train_i, "\ttest_i", test_i, "\n")
```
