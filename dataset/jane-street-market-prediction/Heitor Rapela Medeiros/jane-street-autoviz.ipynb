## Import Libs

```python
import pandas as pd
from autoviz.AutoViz_Class import AutoViz_Class
```

## Parameters

```python
TRAIN_INPUT = "/kaggle/input/jane-street-market-prediction/train.csv"
TRAIN_NROWS = 100000
AUTO_VIZ_FILENAME = ""
AUTO_VIZ_SEP = ","
AUTO_VIZ_DEP_VAR = "resp"
AUTO_VIZ_HEADER = 0
AUTO_VIZ_VERBOSE = 1
AUTO_VIZ_LOWESS = False
AUTO_VIZ_CHAR_FORMAT = "svg"
AUTO_VIZ_MAX_ROWS_ANALYZED = 1500
AUTO_VIZ_MAX_COLS_ANALYZED = 30
```

## Read train data (100.000 rows)

```python
df = pd.read_csv(TRAIN_INPUT, nrows = TRAIN_NROWS)
df.head(1)
```

## Create autoviz object

```python
AV = AutoViz_Class()
```

## Let's see the magic!

```python
dft = AV.AutoViz(filename=AUTO_VIZ_FILENAME, 
					sep=AUTO_VIZ_SEP, 
					depVar=AUTO_VIZ_DEP_VAR, 
					dfte=df, 
					header=AUTO_VIZ_HEADER, 
					verbose=AUTO_VIZ_VERBOSE,
                    lowess=AUTO_VIZ_LOWESS,
                    chart_format=AUTO_VIZ_CHAR_FORMAT,
                    max_rows_analyzed=AUTO_VIZ_MAX_ROWS_ANALYZED,
                    max_cols_analyzed=AUTO_VIZ_MAX_COLS_ANALYZED)
```
