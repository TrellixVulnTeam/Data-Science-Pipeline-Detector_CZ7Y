# <p style="background-color:skyblue; font-family:newtimeroman; font-size:150%; text-align:center">2. Import Libs </p> 

```python
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import sweetviz as sv
from IPython.display import IFrame
```

<a id='3'></a>
# <p style="background-color:skyblue; font-family:newtimeroman; font-size:150%; text-align:center">3. Parameters </p> 

```python
TRAIN_INPUT = "/kaggle/input/jane-street-market-prediction/train.csv"
TRAIN_NROWS = 100000
IFRAME_SRC = "test.html"
IFRAME_WIDTH = 1920
IFRAME_HEIGHT = 1080
```

<a id='4'></a>
# <p style="background-color:skyblue; font-family:newtimeroman; font-size:150%; text-align:center">4. Visualizations </p> 

```python
df = pd.read_csv(TRAIN_INPUT, nrows = TRAIN_NROWS)
df.head(1)
```

```python
analyze = sv.analyze(df)
analyze.show_html(IFRAME_SRC)
```

```python
IFrame(src=IFRAME_SRC, width=IFRAME_WIDTH, height=IFRAME_HEIGHT)
```
