```python
train = train.query('date > 85').reset_index(drop = True)   # 只保留第86天及以后的data
```

```python
x_train=train.iloc[:,7:]
x_train=x_train.iloc[:,:-1]
corr=x_train.corr()
```

```python
mask_corr=np.triu(np.ones_like(corr,dtype=bool))
corr=corr.mask(mask_corr)
to_drop=[c for c in corr.columns if any(corr[c]>0.95)]
df = train.drop(to_drop,axis=1)
x_train=x_train.drop(to_drop,axis=1)
```

```python
x_train.fillna(x_train.mean(),inplace=True)
```
