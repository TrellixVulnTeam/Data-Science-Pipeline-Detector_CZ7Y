purpose of this notebook is to check the relationships between features with the same tag.

````python
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
````

````python
import pandas as pd
train = pd.read_csv("../input/jane-street-market-prediction/train.csv")
test = pd.read_csv("../input/jane-street-market-prediction/example_test.csv")
features = pd.read_csv("../input/jane-street-market-prediction/features.csv")
submission = pd.read_csv("../input/jane-street-market-prediction/example_sample_submission.csv")
````

# Check feature.csv

````python
plt.subplots(figsize=(20, 10))
sns.heatmap(features.set_index("feature"))
````

````python
plt.subplots(figsize=(20, 10))
tag_count = features.set_index("feature").sum(axis=1).value_counts().sort_index().to_frame()
plt.bar(tag_count.index, tag_count[0])
for x, y in zip(tag_count.index, tag_count[0]):
    plt.text(x, y, y ,ha='center', va='bottom')
plt.title("tag count by feature")
````

no tag related with feature_0

## tag_0 ~ tag_5

````python
plt.subplots(figsize=(20, 10))
tag_count = features.set_index("feature")[["tag_0", "tag_1", "tag_2", "tag_3", "tag_4", "tag_5"]].sum(axis=1).value_counts().sort_index().to_frame()
plt.bar(tag_count.index, tag_count[0])
for x, y in zip(tag_count.index, tag_count[0]):
    plt.text(x, y, y ,ha='center', va='bottom')
plt.title("tag_0 ~ tag_5 count by feature")
````

There are no duplicate tags_0 to tag_5

Check the tags with the same tag except tag_0 to tag_5

````python
check_the_features_list = [
    ["feature_7", "feature_9", "feature_11", "feature_13", "feature_15"],    # tag_6, tag_9, tag_11
    ["feature_8", "feature_10", "feature_12", "feature_14", "feature_16"],   # tag_6, not tag_9, tag_11
    ["feature_17", "feature_19", "feature_21", "feature_23", "feature_25"],  # tag_6, tag_9, tag_12
    ["feature_18", "feature_20", "feature_22", "feature_24", "feature_26"],  # tag_6, not tag_9, tag_12
    ["feature_27", "feature_29", "feature_31", "feature_33", "feature_35"],  # tag_6, tag_9, tag_13
    ["feature_28", "feature_30", "feature_32", "feature_34", "feature_36"]   # tag_6, not tag_9, tag_13
]

check_feature_tag = [
    "tag_6, tag_9, tag_11",
    "tag_6, not tag_9, tag_11",
    "tag_6, tag_9, tag_12",
    "tag_6, not tag_9, tag_12",
    "tag_6, tag_9, tag_13",
    "tag_6, not tag_9, tag_13"
]
````

````python
fig, axes = plt.subplots(2, 3, figsize=(20, 10))
for n, (feature, tag) in enumerate(zip(check_the_features_list, check_feature_tag)):
    c = n % 2
    v = n // 2
    sns.heatmap(train[feature].corr(), annot=True, ax=axes[c, v])
    axes[c, v].set_title(tag)
````

The features related to tag_11 are not so correlated due to the difference between tags0 to tag_5.

But The features related to tag_12 or tag_13 is correlated

## tag_6 or tag_9

````python
plt.subplots(figsize=(20, 10))
tag_count = features.set_index("feature")[["tag_6", "tag_9"]].sum(axis=1).value_counts().sort_index().to_frame()
plt.bar(tag_count.index, tag_count[0])
for x, y in zip(tag_count.index, tag_count[0]):
    plt.text(x, y, y ,ha='center', va='bottom')
plt.title("tag_6 or tag_9 count by feature")
````

````python
fig, axes = plt.subplots(1, 2, figsize=(20, 10))
features[
    features["tag_6"] &
    features["tag_9"]
].set_index("feature").sum().plot(kind="barh", ax=axes[0])
axes[0].set_title("tag_6 and tag_9")

features[
    features["tag_6"] &
    ~features["tag_9"]
].set_index("feature").sum().plot(kind="barh", ax=axes[1])
axes[1].set_title("tag_6 and not tag_9")
````
