```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

import os
```

```python
os.listdir("../input/jane-street-market-prediction")
```

```python
train = pd.read_csv("../input/jane-street-market-prediction/train.csv")
test = pd.read_csv("../input/jane-street-market-prediction/example_test.csv")
features = pd.read_csv("../input/jane-street-market-prediction/features.csv")
```

```python
FEATURES = [f"feature_{N}" for N in range(130)]
RESP = [f"resp_{N}" for N in range(1, 5)] + ["resp"]
TS_ID = ["ts_id"]
DATE = ["date"]
```

# Per Date

## Transaction

```python
agg_per_date = pd.concat([
    train.groupby(DATE)["ts_id"].count(),
    train.groupby(DATE)["resp_1"].mean(),
    train.groupby(DATE)["resp_2"].mean(),
    train.groupby(DATE)["resp_3"].mean(),
    train.groupby(DATE)["resp_4"].mean(),
    train.groupby(DATE)["resp"].mean()
], axis=1)
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
agg_per_date["ts_id"].plot(ax=ax)
plt.title("Transaction Count Per Date")
plt.show()
```

## Reception

```python
fig, ax = plt.subplots(figsize=(20, 10))
agg_per_date[["resp_1", "resp_2", "resp_3", "resp_4", "resp"]].plot(ax=ax)
plt.title("Reception Mean Per Date")
plt.legend()
plt.show()
```

Reception Correlogram

```python
from pandas.plotting import autocorrelation_plot
plt.subplots(figsize=(20, 10))
autocorrelation_plot(agg_per_date["resp"])
```

```python
fig, axes = plt.subplots(5, 2, figsize=(20, 50))
sns.regplot(x="resp_1", y="resp_2", data=agg_per_date, ax=axes[0, 0])
sns.regplot(x="resp_1", y="resp_3", data=agg_per_date, ax=axes[0, 1])
sns.regplot(x="resp_1", y="resp_4", data=agg_per_date, ax=axes[1, 0])
sns.regplot(x="resp_1", y="resp", data=agg_per_date, ax=axes[1, 1])
sns.regplot(x="resp_2", y="resp_3", data=agg_per_date, ax=axes[2, 0])
sns.regplot(x="resp_2", y="resp_4", data=agg_per_date, ax=axes[2, 1])
sns.regplot(x="resp_2", y="resp", data=agg_per_date, ax=axes[3, 0])
sns.regplot(x="resp_3", y="resp_4", data=agg_per_date, ax=axes[3, 1])
sns.regplot(x="resp_3", y="resp", data=agg_per_date, ax=axes[4, 0])
sns.regplot(x="resp_4", y="resp", data=agg_per_date, ax=axes[4, 1])
```
