# Plotting All Features

In this competition we are provided an anonymized set of features.

Even though features are anonymized, we can plot them to gain an understainding of their distributions.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
train = pd.read_csv('../input/jane-street-market-prediction/train.csv')
test = pd.read_csv('../input/jane-street-market-prediction/example_test.csv')
ss = pd.read_csv('../input/jane-street-market-prediction/example_sample_submission.csv')
```

# Target Distributions
- Response, and weights

```python
fig, axs = plt.subplots(2, 3, figsize=(12, 10))
axs = axs.flatten()
sns.distplot(train['resp'], ax=axs[0])
axs[0].set_title('resp')
sns.distplot(train['weight'], ax=axs[1])
axs[1].set_title('weight')
sns.distplot(train['resp_1'], ax=axs[2])
sns.distplot(train['resp_2'], ax=axs[3])
sns.distplot(train['resp_3'], ax=axs[4])
sns.distplot(train['resp_4'], ax=axs[5])
```

```python
FEAT_COLS = [c for c in train.columns if 'feature' in c]
for f in FEAT_COLS:
    fig, axs = plt.subplots(1, 4, figsize=(15, 4))
    sns.distplot(train[f], ax=axs[0])
    sns.distplot(train.query('weight > 0')[f], ax=axs[1])
    try:
        sns.distplot(train.query('weight > 0 and resp > 0')[f].dropna().apply(np.log1p), ax=axs[2])
        sns.distplot(train.query('weight > 0 and resp < 0')[f].dropna().apply(np.log1p), ax=axs[2])
    except:
        pass
    train.sample(5000).plot(kind='scatter', x=f, y='resp', ax=axs[3])
    fig.suptitle(f, fontsize=15, y=1.1)
    
    axs[0].set_title('feature distribution')
    axs[1].set_title('only weight > 0')
    axs[2].set_title('log transform')
    axs[3].set_title('feature vs. response')
    
    plt.tight_layout()
    plt.show()
```
