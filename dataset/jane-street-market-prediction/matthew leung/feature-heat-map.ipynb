```python
import pandas as pd
feature_tag = pd.read_csv("/kaggle/input/jane-street-market-prediction/features.csv")
feature_tag
```

```python
tags=[x for x in feature_tag.columns if x.startswith('tag')]
for t in tags:
    feature_tag[t] = feature_tag[t].apply(lambda x: 1 if x else 0)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
n = len(feature_tag['feature'])
m = np.zeros((n,n))
for i,fi in enumerate(feature_tag['feature']):
    for j,fj in enumerate(feature_tag['feature']):
        if i == j:
            m[i,j] = 0
        else:
            vi = feature_tag[feature_tag.feature == fi][tags].values
            vj = feature_tag[feature_tag.feature == fj][tags].values
            #print(vi,vj)
            c = cosine_similarity(vi,vj)
            #print(c[0,0])
            m[i,j] = c[0,0]
```

```python
import seaborn as sns
sns.heatmap(m)
```
