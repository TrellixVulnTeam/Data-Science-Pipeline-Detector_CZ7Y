```python
import pandas as pd
from tqdm.auto import tqdm
```

```python
train_df=pd.read_csv('../input/jane-street-market-prediction/train.csv',nrows=100)
```

```python
features=[f'feature_{i}' for i in range(130)]
```

```python
means=train_df[features].mean()
```

```python
import janestreet
env = janestreet.make_env() # initialize the environment
iter_test = env.iter_test() # an iterator which loops over the test set

for (test_df, pred_df) in tqdm(iter_test):
    test_df[features].fillna(means)
    env.predict(pred_df)
```
