Here, I try to use real short position data to determin typical time horizons. 

```python
#I use the Norwegian Financial Authorities data which is updated daily. Two years of data.
data = pd.read_csv('https://ssr.finanstilsynet.no/api/v2/instruments/export-csv',delimiter=';',dayfirst=True)
```

```python
data['ShortPercent'] = data['ShortPercent'].apply(lambda x: x.replace(',','.')).astype(float)
data = data.sort_values(by=['Date'])
```

```python
data = data.sort_values(by=['Date'])
data.groupby('Name').count()['Shares'].hist()
```

```python
data[data['Name']=='AKER SOLUTIONS']
```

Plotting the autocorrelation for `Subsea 7` and `Aker Solutions`. These are both offshore service companies. 

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

d = data[data['Name'] == 'SUBSEA 7']
plot_acf(d['Shares'], lags=40,title='Subsea 7')
d = data[data['Name'] == 'AKER SOLUTIONS']
plot_acf(d['Shares'], lags=40,title='Aker Solutions')
```

For `SUBSEA 7` the horizon seems to be 5 days while `Aker Solutions` shorts have a more sporadic pattern which in general is shorter.
