This notebook will contain some analysis about the objective function we need to optmize in this competition. In the first part we will analyse the behavior of the variable t. First, lets write a function to calculate it.

```python
def calc_t(p_vec):
    sp = np.sum(p_vec)
    sqrp = np.sqrt(np.sum(p_vec**2)*len(p_vec)/250)
    t = sp/sqrp
    return(t)
```

In the first analysis we will keep the sum of p and size of te vector fixed

```python
def gen_vec(size, sum_val, offset):
    p_vec = np.random.uniform(size=size)
    p_vec = p_vec*sum_val/np.sum(p_vec) - offset
    return(p_vec)
        
```

```python
gen_vec(10, 10, 0)
```

```python
std_vec = []
t_vec = []
for i in range(10000):
    p_vec = gen_vec(10, 20, 0)
    std_vec.append(np.std(p_vec))
    t_vec.append(calc_t(p_vec))
```

```python
plt.scatter(std_vec, t_vec)
plt.show()
```

```python
tm_vec = []
for j in range(100):
    t_vec = []
    for i in range(100):
        p_vec = gen_vec(10 + j, 20, 0)
        t_vec.append(calc_t(p_vec))
    tm_vec.append(np.mean(t_vec))
```

```python
plt.scatter([(10 + j) for j in range(100)], tm_vec)
plt.show()
```

```python
tm_vec = []
for j in range(100):
    t_vec = []
    for i in range(100):
        p_vec = gen_vec(10, 20 + j, 0)
        t_vec.append(calc_t(p_vec))
    tm_vec.append(np.mean(t_vec))
```

```python
plt.scatter([(20 + j) for j in range(100)], tm_vec)
plt.show()
```

```python
tm_vec = []
stdm_vec=[]
for j in range(100):
    t_vec = []
    std_vec=[]
    for i in range(100):
        p_vec = gen_vec(10, 20, j/100)
        t_vec.append(calc_t(p_vec))
        std_vec.append((p_vec < 0).sum())
    tm_vec.append(np.mean(t_vec))
    stdm_vec.append(np.mean(std_vec))
```

```python
plt.scatter([(j/500) for j in range(100)], tm_vec)
plt.show()
```

```python
plt.scatter([(j/500) for j in range(100)], stdm_vec)
plt.show()
```

```python
plt.scatter(stdm_vec, tm_vec)
plt.show()
```

So the function penalizes instability and losses. In general stable and less risky algorithms should perform better in this task.
