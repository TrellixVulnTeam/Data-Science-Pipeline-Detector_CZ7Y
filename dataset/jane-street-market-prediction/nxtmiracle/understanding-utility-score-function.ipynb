### (2). 

In finance, volatility (usually denoted by $Ïƒ$) is the degree of variation of a trading price series over time, usually measured by the standard deviation of logarithmic returns.

Mathematically, the generalized volatility $\sigma_{T}$ for time horizon $T$ in years is expressed as:
$$
\sigma_{T} = \sigma_{annually}\sqrt{T}
$$
Therefore, if the daily logarithmic returns of a stock have a standard deviation of $\sigma_{daily}$ and the time period of returns is $P$ in trading days, the **annualized volatility** is
$$
\sigma_{P} = \sigma_{daily}\sqrt{P}
$$
A common assumption is that $P = 252$ trading days in any given year. Then, if $\sigma_{daily}$ = 0.01, the annualized volatility is

$$
{\displaystyle \sigma _{\text{annually}}=0.01{\sqrt {252}}=0.1587.}
$$
The monthly volatility (i.e., T = 1/12 of a year or P = 252/12 = 21 trading days) would be
$$
{\displaystyle \sigma _{\text{monthly}}=0.1587{\sqrt {\tfrac {1}{12}}}=0.0458.}
\\
{\displaystyle \sigma _{\text{monthly}}=0.01{\sqrt {\tfrac {252}{12}}}=0.0458.}
$$
The annualised daily volatility is
$$
\sigma_{annualized\_daily} = \sigma_{daily}\sqrt{\frac{250}{n}}
$$

### (3).

If we take $p_{i}$ as the logarithmic return of the i-th day and estimate the mean logarithmic return is 0,
then we have 
$$
\sigma_{p} = \sqrt{\frac{\sum{(X-\bar{X}^2)}}{n}} = \sqrt{\frac{\sum{(p_i - 0)^2}}{n}} = \frac{\sqrt{\sum{(p_i^2)}}}{\sqrt{n}}
$$

The logarithmic return is 
$$
R_{log} = ln(\frac{V_f}{V_i})
$$
where $V_{f}$ is the final value and $V_{i}$ is the initial value.
Thus, $p_{i} = 0$ means ${V_f}={V_i}$, i.e., the return is 0.

### (4).

Sharpe Ratio is another finacial metric
$$
Sharpe Ratio = \frac{R_{p}-R_{f}}{\sigma_{p}}
$$
$R_{p}$ is the investment return and $R_{f}$ is the risk-free return.

Similarily, we estimate the risk-free return $R_f=0$ and we have
$$
Sharpe Ratio = \frac{R_{p}-R_{f}}{\sigma_{p}} = \frac{\sum{p_{i}}}{\sigma_{p}} = \frac{\sum{p_{i}}}{\sqrt{\sum{(p_i^2)}}}*\sqrt{n}
$$

### (5).
Recall that
$$
\mu_{p} = \frac{\sum{p_{i}}}{n}
$$
So we can rewrite $t$ as 
\begin{align*}
t
&=
\frac{\sum{p_{i}}}{\sqrt{\sum{p_{i}^{2}}}}*\sqrt{\frac{250}{n}}
\\
&=
\frac{\sum{p_{i}}}{n}*\frac{n}{\sqrt{\sum{p_{i}^{2}}}}*\sqrt{\frac{250}{n}}
\\
&=
\frac{\sum{p_{i}}}{n}*\frac{\sqrt{n}}{\sqrt{\sum{p_{i}^{2}}}}*\sqrt{250}
\\
&=
\frac{\mu_{p}}{\sigma_{p}}*\sqrt{250}
\end{align*}
Since $\frac{\mu_{p}}{\sigma_{p}} = \frac{Sharpe Ratio}{n}$,
so we can call it as Mean Sharpe Ratio (MSR).
$$
t = MSR*\sqrt{250}
$$
Similar to the definition of the annualized volatility in (2),
**$t$ can be called Annualized Mean Sharpe Ratio (AMSR)**.
