Why anonymized?
- Sometimes the data is classfied. But the host wants people outside to utilize the data without important business related secrets revealed. So they hash or hide the names and/or values. 

What to do?
- Guess: guess the meaning and type of the columns
- Explore: find relationships within feature pairs, find feature groups
