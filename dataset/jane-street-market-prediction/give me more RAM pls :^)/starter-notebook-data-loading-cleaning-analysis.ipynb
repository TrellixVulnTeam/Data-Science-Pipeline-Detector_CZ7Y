For exploration purposes we limit the number of rows to 500k (2.3M rows total). This can be changed through the "nrows" argument.

```python
train = pd.read_csv("../input/jane-street-market-prediction/train.csv", nrows=500000)
```

```python
train.drop(['weight','resp_1','resp_2','resp_3','resp_4'],axis=1,inplace=True)
```

```python
gc.collect()
```

```python
X_train = train[train.index < 400000].drop(['resp','date','ts_id'],axis=1)
y_train = train[train.index < 400000]['resp']
X_test = train[train.index > 400000].drop(['resp','date','ts_id'],axis=1)
y_test = train[train.index > 400000]['resp']
```

```python
X_train
```

```python
X_train.fillna(0,inplace=True)
X_test.fillna(0,inplace=True)
```

```python
gc.collect()
```

Here we fit a linear regression.

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train,y_train)
```

```python
reg.score(X_test,y_test)
```
