# 한번에 실행

```python
import time
from selenium import webdriver
import pandas as pd
import requests
from bs4 import BeautifulSoup
from selenium.webdriver.common.keys import Keys

url = 'https://search.naver.com/search.naver?where=news&sm=tab_jum&query='
search = '//*[@id="nx_query"]'
pagelist = '//*[@id="main_pack"]/section/div/div[3]/ul'
classname = 'bx'
nextpage = '//*[@id="main_pack"]/div[2]/div/a[2]'

# 네이버 접속하기
driver = webdriver.Chrome('chromedriver')

# 검색어 
words = ['안철수','문재인','이재명']
mylist = []

# 반복
pagelimit = 2
for k in range(len(words)):
    driver.get(url+words[k])
    time.sleep(3)
        
    for j in range(pagelimit):
        src1 = driver.find_element_by_xpath(pagelist) # 뉴스 리스트
        src2 = src1.find_elements_by_class_name(classname) # 뉴스 리스트 전체항목
        
        for i in range(len(src2)): # 뉴스리스트 1~10개
            
            print('단어:{}, 페이지수 {}/{}'.format(words[k],j+1,pagelimit))
            rst1 = words[k] # 검색어
            rst2 = src2[i].find_element_by_xpath('div/div/a').text # 제목
            rst3 = src2[i].find_elements_by_tag_name("a")[0].get_attribute('href')

            mylist.append([rst1,rst2,rst3])
                 
        driver.find_element_by_xpath(nextpage).click()

df = pd.DataFrame(mylist)
df.columns = ['검색어','제목','url']
df.to_csv('news_crawl.csv',index=False,encoding='utf-8-sig')
```
