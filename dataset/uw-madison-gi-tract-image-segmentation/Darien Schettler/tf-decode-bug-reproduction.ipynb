```python
import numpy as np
import pandas as pd
import tensorflow as tf
import cv2
from tqdm.notebook import tqdm
```

```python
# Load our dummy model
dummy_model = tf.keras.applications.ResNet50()
```

```python
# Make a dataset of 250000 smaller images (similar to dataset in uwm comp)
for i in tqdm(range(250000), total=250000):
    cv2.imwrite(f"test_image_{i}.png", np.ones((224,224,3), dtype=np.uint8))
```

```python
# Here's the warning FYI
tf.io.decode_image(tf.io.read_file(f"test_image_1.png")).shape
```

```python
# Make our pipeline
def tf_load_img(img_path):
    img = tf.io.read_file(img_path)
    img = tf.image.decode_image(img)
    return img

dummy_ds = tf.data.Dataset.from_tensor_slices(tf.io.gfile.glob("/kaggle/working/*.png"))
dummy_ds = dummy_ds.map(tf_load_img)
dummy_ds = dummy_ds.batch(16)
```

```python
dummy_preds = dummy_model.predict(dummy_ds, verbose=1)
```
