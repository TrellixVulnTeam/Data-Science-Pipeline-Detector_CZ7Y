## How to create VIDEO files from frame sequences

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import ast

from tqdm.notebook import tqdm
tqdm.pandas()
```

#### Load the train file


```python
PATH = '../input/tensorflow-great-barrier-reef/train_images/video_'
    
def filepaths(row):
    v = str(row.video_id)
    f = str(row.video_frame)
    row['filepath'] = PATH+v+'/'+f+'.jpg'
    return row

train = pd.read_csv('../input/tensorflow-great-barrier-reef/train.csv')
train = train.progress_apply(filepaths, axis =1)

train['len_ann'] = [ len(a) for a in train.annotations ]

train.head()
```

### Verify the length of sequences

In video 1, sequence 18048, is the shorted video with 71 frames..

```python
for video in train.video_id.unique():
    for seq in train[train.video_id==video].sequence.unique():
        print('Video:', video, '| seq:', seq,'| len:', len(train[train.sequence==seq]  ) )        
```

```python
# Shortest Video
# Video: 1 | seq: 18048 | len: 71
video_short = train[(train.video_id ==1) & (train.sequence==18048)]
video_short.head()
```

Check if the rectangle is properly placed in the image.

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import ast

i = 0

rec_dict=ast.literal_eval(video_short.iloc[i].annotations[1:-1])
x,y,w,h=rec_dict['x'],rec_dict['y'],rec_dict['width'],rec_dict['height']

fig, ax = plt.subplots(figsize=(20,10))
img = cv2.imread(video_short.filepath.tolist()[i], -1)
img = cv2.rectangle(img,(x,y),(x+w,y+h), (252, 3, 3), 2 )

plt.imshow(img)
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

Following, create a video ('*project.avi*') from the sequence above (18048) where the rectangles are drawn in the video. 



```python
import cv2 
import numpy as np
import ast

img_array = []

for i, filename in enumerate(video_short.filepath.tolist()):

    img = cv2.imread(filename)
    rec_dict=ast.literal_eval(video_short.iloc[i].annotations) #[1:-1])
    
    for j in range(len(rec_dict)):
        x,y,w,h=rec_dict[j]['x'],rec_dict[j]['y'],rec_dict[j]['width'],rec_dict[j]['height']

        img = cv2.rectangle(img,(x,y),(x+w,y+h), (252, 3, 3), 2 )
    
    height, width, layers = img.shape
    size = (width,height)
    img_array.append(img)

out = cv2.VideoWriter('project.avi',cv2.VideoWriter_fourcc(*'DIVX'), 15, size)
 
for i in range(len(img_array)):
    out.write(img_array[i])
out.release()    
    
```

You can download the video file from output folder and watch the sequence in your computer.
