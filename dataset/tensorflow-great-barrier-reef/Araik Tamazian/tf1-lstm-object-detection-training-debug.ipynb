```python
from string import Template
import tensorflow as tf
import os
```

```python
!git clone https://github.com/tensorflow/models/
    
# Check out a certain commit to ensure that future changes in the TF ODT API codebase won't affect this notebook.
!cd models && git checkout ac8d06519
```

```python
%%bash
cd models/research

# Compile protos.
protoc object_detection/protos/*.proto --python_out=.

# Install TensorFlow Object Detection API.
cp object_detection/packages/tf1/setup.py .
pip install -q --user .

# Test if the Object Dectection API is working correctly
python object_detection/builders/model_builder_test.py
```

```python
%%writefile models/research/setup.py

"""Setup script for object_detection with TF2.0."""
import os
from setuptools import find_packages
from setuptools import setup

setup(
    name='lstm_object_detection',
    version='0.1',
    install_requires=[],
    include_package_data=True,
    packages=(
        [p for p in find_packages() if p.startswith('lstm_object_detection')] +
        find_packages(where=os.path.join('.', 'slim'))),
    package_dir={
        'datasets': os.path.join('slim', 'datasets'),
        'nets': os.path.join('slim', 'nets'),
        'preprocessing': os.path.join('slim', 'preprocessing'),
        'deployment': os.path.join('slim', 'deployment'),
        'scripts': os.path.join('slim', 'scripts'),
    },
    description='Tensorflow LSTM Object Detection Library',
    python_requires='>3.6',
)
```

```python
%%bash
cd models/research

# Compile protos.
protoc lstm_object_detection/protos/*.proto --python_out=.

# Install TensorFlow Object Detection API.
# When Kaggle notebook upgrade to TF 2.7, you can use the default setup.py script:
# cp object_detection/packages/tf2/setup.py .
pip install -q --user .

# Test if the Object Dectection API is working correctly
python lstm_object_detection/model_builder_test.py
```
