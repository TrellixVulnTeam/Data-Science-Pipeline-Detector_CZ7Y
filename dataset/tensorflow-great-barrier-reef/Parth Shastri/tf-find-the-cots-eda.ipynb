```python
import pandas as pd
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import cv2
```

```python
input_csv = "../input/tensorflow-great-barrier-reef/train.csv"
train_df = pd.read_csv(input_csv)
train_df.head()
```

```python
train_pos_df = train_df[train_df["annotations"]!="[]"]  # select only the positive samples for training
print(f"The no if positive examples :{len(train_pos_df)}")
print(f"The number of negative examples :{len(train_df) - len(train_pos_df)}")
```

```python
image_dir = "../input/tensorflow-great-barrier-reef/train_images"
train_pos_df["annotations"] = train_pos_df["annotations"].map(eval)
ex_paths = [f"{image_dir}/video_{int(id)}/{int(frame)}.jpg" 
            for id, frame in zip(train_pos_df["video_id"], train_pos_df["video_frame"])]
train_pos_df["img_paths"] = ex_paths
train_pos_df.head()
```

```python
nrows, ncols = 2, 2
fig, axs = plt.subplots(nrows=nrows, ncols=ncols, constrained_layout=True)

show_paths = np.reshape(list(train_pos_df["img_paths"])[:4], (2, 2))
for i in range(nrows):
    for j in range(ncols):
        img = Image.open(show_paths[i, j]).convert("RGB")
        img_array = np.asarray(img)
        annots = train_pos_df[train_pos_df["img_paths"] == show_paths[i, j]]["annotations"]
        if annots is not None:
            for annot in list(annots)[0]:
                x_min, y_min = annot["x"], annot["y"]
                x_max, y_max = annot["x"] + annot["width"], annot["y"] + annot["height"]
                img_array = cv2.rectangle(img_array, (x_min, y_min), (x_max, y_max), (255, 0, 0), 4)
        axs[i, j].imshow(img_array)
        print(f"image size: {img_array.shape}")
fig.show()
```
