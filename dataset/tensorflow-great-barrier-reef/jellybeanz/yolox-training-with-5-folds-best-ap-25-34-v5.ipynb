```python
%%writefile cots_config.py

import os
from yolox.exp import Exp as MyExp

class Exp(MyExp):
    def __init__(self):
        super(Exp, self).__init__()
        self.depth = 0.33
        self.width = 0.50
        self.exp_name = os.path.split(os.path.realpath(__file__))[1].split(".")[0]
        
        # dirs
        self.data_dir = '/kaggle/input/cotscocostratifiedk5folds-4919-imgs/dataset/images'
        self.train_ann = '/kaggle/input/cotscocostratifiedk5folds-4919-imgs/dataset/images/annotations/train.json'
        self.val_ann = '/kaggle/input/cotscocostratifiedk5folds-4919-imgs/dataset/images/annotations/valid.json'
        
        self.input_size = (960, 960)
        self.test_size = (960, 960)
        self.num_classes = 1
        
        self.max_epoch = 37
        self.data_num_workers = 2
        self.eval_interval = 1
        
        # Augmentations
        self.mosaic_prob = 1.0
        self.mosaic_scale = (0.5, 1.5)
        self.mixup_prob = 1.0
        self.hsv_prob = 1.0
        self.flip_prob = 0.5
        self.no_aug_epochs = 2
```
