```python
import pandas as pd 
import tensorflow as tf 
import numpy as np
import torch 
import pytorch_lightning as pl
import matplotlib.pyplot as plt
import torchvision

def read_csv(path):
    df=pd.read_csv(path)
    return df

def read_img(path):
    img=torchvision.io.decode_jpeg(path)
    return img

def plot_batch(img_paths,r,c):
    _,axs=plt.subplots(r,c,figsize=(10,10))
    axs=axs.flatten()
    for n, ax in enumerate(axs):
        ax.imshow(read_img(img_paths[n]))
        
    plt.tight_layout()
        
    plt.show()
```
