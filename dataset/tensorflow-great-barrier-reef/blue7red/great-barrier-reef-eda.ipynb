# get folder name => Target 

```python
def getLastFolderName(x): 
    pathwithoutFilename = os.path.split(x)[0] 
    lastFolderName = pathwithoutFilename.split("/")[-1] 
    lastFolderName = lastFolderName.replace("video_","")
    return lastFolderName

def getFileName(x): 
    pathwithoutFilename = os.path.split(x)[1] 
    lastFolderName = pathwithoutFilename.split(".")[0] 
    return lastFolderName


def getImageId(x): 
    imageId = getLastFolderName(x) + "-" + getFileName(x)
    return imageId

filePaths = filepaths.copy()
fileNames = list(map(lambda x: getFileName(x),filepaths)) 
fileTypes = list(map(lambda x: getLastFolderName(x),filepaths)) 
image_ids = list(map(lambda x: getImageId(x),filepaths)) 
print(fileTypes[:5])
print(fileNames[:5])
print(filepaths[:5])
print(image_ids[:5])
```
