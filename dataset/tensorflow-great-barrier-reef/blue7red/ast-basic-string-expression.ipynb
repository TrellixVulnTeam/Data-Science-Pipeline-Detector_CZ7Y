```python
import pandas as pd
import ast
```

# String => Expression

```python
expressionString1 = "{'a': 1, 'b': 2, 'c': 3}"
print(ast.literal_eval(expressionString1))

expressionString2 = "[1, 2, 3, 4, 5]"
print(ast.literal_eval(expressionString2))

expressionString3 = "[(1,2),(3,4),(5,6)]"
print(ast.literal_eval(expressionString3))

expressionString4 = "[{'x': 559, 'y': 213, 'width': 50, 'height': 32}]"
print( ast.literal_eval(expressionString4) )

expressionString5 = "[{'x': 100, 'y': 100, 'width': 50, 'height': 50},{'x': 200, 'y': 200, 'width': 30, 'height': 30}]"
print( ast.literal_eval(expressionString5) )
```
