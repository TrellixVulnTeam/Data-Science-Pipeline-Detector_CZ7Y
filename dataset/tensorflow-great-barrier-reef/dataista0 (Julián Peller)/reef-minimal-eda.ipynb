# Check the submission API

## See: [Great Barrier Reef API Tutorial](https://www.kaggle.com/sohier/great-barrier-reef-api-tutorial)

```python
import greatbarrierreef
env = greatbarrierreef.make_env()   # initialize the environment
iter_test = env.iter_test()    # an iterator which loops over the test set and sample submission
```

```python
for (pixel_array, df_pred) in iter_test:  # iterate through all test set images
    print(f"Pixels shape: {pixel_array.shape}")
    print(f"df_pred shape: {df_pred.shape}")
    
    img = PIL.Image.fromarray(pixel_array).resize((400, 256))
    display(img)
    
    display(df_pred.head())
    df_pred['annotations'] = '0.5 123 456 10 10'  # make your predictions here
    env.predict(df_pred)   # register your predictions
```
