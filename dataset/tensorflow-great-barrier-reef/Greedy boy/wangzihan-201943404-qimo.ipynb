```python
def draw_box(num0):
    
    impath=impaths[num0]
    image=cv2.imread(impath)
    #H,W=image.shape[0],image.shape[1]
    #file=impath[0:-4].split('/')[-1]
    #print(file, DICT0[num0])

    item = DICT0[num0]       
    if item!={} and type(item)==dict:
        x=item['x']
        y=item['y']
        w=item['width'] 
        h=item['height']
        x0=int(x-w/2)
        y0=int(y-h/2)
        x1=int(x+w/2)
        y1=int(y+h/2)

        cv2.rectangle(image,(x0,y0),(x1,y1),(0,0,255),4) 
        #print('dict')

    elif item!={} and type(item)==tuple:
        for item2 in item:
            x=item2['x']
            y=item2['y']
            w=item2['width'] 
            h=item2['height']
            x0=int(x-w/2)
            y0=int(y-h/2)
            x1=int(x+w/2)
            y1=int(y+h/2)

            cv2.rectangle(image,(x0,y0),(x1,y1),(0,0,255),4)
            #print('tuple')

        #cv2.imwrite('/kaggle/working/source0/'+file+'.png',image)        
        #plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
        #plt.show() 
        
    return image
```

```python
images1=[]
#for i in tqdm(range(len(impaths))):
for i in tqdm(range(200)):
    images1+=[draw_box(i)]
```

### 动画

```python
from matplotlib import animation, rc
rc('animation', html='jshtml')
```

```python
def create_animation(ims):  
    fig=plt.figure(figsize=(12,6))
    plt.axis('off')
    im=plt.imshow(cv2.cvtColor(ims[0],cv2.COLOR_BGR2RGB))
    
    def animate_func(i):
        im.set_array(cv2.cvtColor(ims[i],cv2.COLOR_BGR2RGB))
        return [im]

    return animation.FuncAnimation(fig, animate_func, frames=len(ims), 
                                    interval=1000//8)
```

```python
create_animation(images1[20:70])
```
