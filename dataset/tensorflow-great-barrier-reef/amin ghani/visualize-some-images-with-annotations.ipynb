```python
import numpy as np
import pandas as pd
train = pd.read_csv('/kaggle/input/tensorflow-great-barrier-reef/train.csv')
```

```python
train
```

```python
import matplotlib.pyplot as plt
import cv2
import ast
```

```python
def show_images_with_annotation(video_number, num_fig, start_frame):
    images = []
    fig = plt.figure(figsize=(16, 14))
    for i in range(num_fig):
        BASE = '/kaggle/input/tensorflow-great-barrier-reef/train_images/'
        img = cv2.imread(BASE+'video_'+str(video_number)+'/'+str(i+start_frame)+'.jpg')
        x = ast.literal_eval(train[train.video_id == video_number].iloc[i+start_frame,:]['annotations'])
        for l in x:
            image = cv2.rectangle(img, (l['x'],l['y']) , (l['x']+l['width'],l['y']+l['height']) ,color=(255,0,0), thickness=5)
        images.append(image)

    rows=3
    columns=3
    for i in range(num_fig):
        fig.add_subplot(rows, columns, i+1)
        plt.imshow(images[i])
        plt.axis('off')
```

```python
show_images_with_annotation(0, 9, 50)
```

```python
show_images_with_annotation(1, 9, 470)
```

```python
show_images_with_annotation(2, 9, 3850)
```
