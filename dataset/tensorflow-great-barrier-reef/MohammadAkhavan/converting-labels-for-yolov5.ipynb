As you can see x and y is pixel format so we need to convert it to Yolov5 annotations format. For this matter we use a library named "ast" that read the annotation in dictionary foramt. then use dictionary values to create labels.

```python
for j in range(len(df)):
    list_anno = ast.literal_eval(df.iloc[j][5])
    for anno in list_anno:
        with open('../yolov5fromat/labels/' + str(df.iloc[j][0]) + '_' + str(df.iloc[j][2]) + '.txt', 'w') as file:
            file.write(str(0) + ' ' + str(anno['x']/width) + ' ' + str(anno['y']/height) + ' ' + str(anno['width']) + ' ' + str(anno['height']) + '\n')
```
