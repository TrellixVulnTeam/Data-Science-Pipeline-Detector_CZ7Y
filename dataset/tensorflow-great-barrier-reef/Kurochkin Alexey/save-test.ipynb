```python
import greatbarrierreef
import pandas as pd
import numpy as np

def predict(image_np):
    predictions = []
    for i in range(2):
        score = np.random.randint(1,10)/10

        x_min = np.random.randint(0,500)
        y_min = 0

        bbox_width = np.random.randint(1,200)
        bbox_height = np.random.randint(1,200)
        
        predictions.append('{:.2f} {} {} {} {}'.format(score, x_min, y_min, bbox_width, bbox_height))
    return ' '.join(predictions)

env = greatbarrierreef.make_env()   # initialize the environment
iter_test = env.iter_test()    # an iterator which loops over the test set and sample submission
submission_dict = {
    'row_num': [],
    'annotations': [],
}

for (image_np, sample_prediction_df) in iter_test:
    sample_prediction_df['annotations'] = predict(image_np)
    env.predict(sample_prediction_df)

    submission_dict['row_num'].append(sample_prediction_df['index'].values[0])
    submission_dict['annotations'].append(sample_prediction_df['annotations'].values[0])
    print('Prediction:', submission_dict['annotations'][-1])
```
