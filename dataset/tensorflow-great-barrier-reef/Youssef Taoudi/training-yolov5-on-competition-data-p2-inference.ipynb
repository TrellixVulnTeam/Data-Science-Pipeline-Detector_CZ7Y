```python
def handle_pred(pred):
    pred.print()
    res = [str(float(conf)) + " " + str(int(xmin)) + " " + str(int(ymin)) + " " + str(int(xmax-xmin)) + " " + str(int(ymax-ymin)) for xmin,ymin,xmax,ymax,conf,_ in pred.xyxy[0]]
    return " ".join(res)
```

```python
import greatbarrierreef
import torch


env = greatbarrierreef.make_env()   # initialize the environment
iter_test = env.iter_test()    # an iterator which loops over the test set and sample submission
print(model_name)

#Loading model
model = torch.hub.load(yolov5_folder, 'custom', path=model_name, force_reload=True, source='local')  # custom model

for pixel_array, sample_prediction_df in iter_test:
    pred = model(pixel_array, size=1280) #Inference
    res = handle_pred(pred) # Annotation formatting
    sample_prediction_df['annotations'] = res
    env.predict(sample_prediction_df)   # register your predictions
```
