```python
def flatten_l_o_l(nested_list):
    """ Flatten a list of lists """
    return [item for sublist in nested_list for item in sublist]

def tf_load_img(img_path, reshape_to=None):
    if reshape_to is None:
        return tf.image.decode_image(tf.io.read_file(img_path), channels=3)
    else:
        return tf.image.resize(tf.image.decode_image(tf.io.read_file(img_path), channels=3), reshape_to)

def get_tl_br(bbox):
    """ Return the top-left and bottom-right bounding box """
    return (bbox['x'], bbox['y']), (bbox['x']+bbox["width"], bbox['y']+bbox["height"])

def plot_image(img_path, annotations=None, **kwargs):
    """ Plot an image and bounding boxes """
    img = np.array(tf_load_img(img_path))
    
    if annotations:
        plt.figure(figsize=(20,10))
        for i, bbox in enumerate(annotations):
            tl_box, br_box = get_tl_br(bbox)
            img = cv2.rectangle(img, tl_box, br_box, (255-2*i,14*i,0), 4)
        plt.imshow(img)
        plt.axis(False)
        plt.title(f"Bounding Boxes Shown in Red ({len(annotations)})", fontweight="bold")
    else:
        plt.figure(figsize=(20,10))
        plt.imshow(img)
        plt.axis(False)
        plt.title("No Bounding Boxes", fontweight="bold")
    plt.tight_layout()
    plt.show()
```

```python
def _bytes_feature(value, is_list=False):
    """Returns a bytes_list from a string / byte."""
    if isinstance(value, type(tf.constant(0))):
        value = value.numpy() # BytesList won't unpack a string from an EagerTensor.
    
    if not is_list:
        value = [value]
    
    return tf.train.Feature(bytes_list=tf.train.BytesList(value=value))


def _float_feature(value, is_list=False):
    """Returns a float_list from a float / double."""
        
    if not is_list:
        value = [value]
        
    return tf.train.Feature(float_list=tf.train.FloatList(value=value))


def _int64_feature(value, is_list=False):
    """Returns an int64_list from a bool / enum / int / uint."""
        
    if not is_list:
        value = [value]
        
    return tf.train.Feature(int64_list=tf.train.Int64List(value=value))


IMG_SHAPE   = (720, 1280, 3)
INPUT_SHAPE = (640,640,3)
MODEL_LEVEL = "d1"
MODEL_NAME  = f"efficientdet-{MODEL_LEVEL}"

N_EPOCH    = 50
BATCH_SIZE = 8

N_TRAIN      = len(sub_train_df)
N_VAL        = len(sub_val_df)
N_EX_PER_REC = 125

CLASS_LABELS = ["COTS",]
N_CLASSES_OD = len(CLASS_LABELS)+1 # Background + Foreground (COTS)

# Whether or not we train from scratch or load
DO_TRAIN=False
PRETRAINED_MODEL_DIR="/kaggle/input/tf-find-cots-keras-efficientdet-implementation/efficientdet-d1-finetune/ckpt-50"

print("\n ... HYPERPARAMETER CONSTANTS ...")
print(f"\t--> MODEL NAME         : {MODEL_NAME}")
print(f"\t--> BATCH SIZE         : {BATCH_SIZE}")
print(f"\t--> # OF EPOCHS        : {N_EPOCH}")
print(f"\t--> IMAGE SHAPE        : {IMG_SHAPE}")
print(f"\t--> INPUT SHAPE        : {INPUT_SHAPE}")
```

```python
def tf_load_image(path, resize_to=INPUT_SHAPE):
    """ Load an image with the correct shape using only TF
    
    Args:
        path (tf.string): Path to the image to be loaded
        resize_to (tuple, optional): Size to reshape image
    
    Returns:
        3 channel tf.Constant image ready for training/inference
    
    """
    
    img_bytes = tf.io.read_file(path)
    img = tf.image.decode_png(img_bytes, channels=resize_to[-1])
    img = tf.image.resize(img, resize_to[:-1])
    img = tf.cast(img, tf.uint8)
    
    return img

def get_scaled_annotations(annotations, original_size=(720,1280), resize_to=(640,640), effdet_return=True):
    """ Get annotations after rescaling """
    
    # Calculate factors to use when identifying new bounding box coordinates
    xr_factor, yr_factor = resize_to[1]/original_size[1], resize_to[0]/original_size[0]
    
    if effdet_return:
        new_annotations = {"xmin":[], "ymin":[], "xmax":[], "ymax":[], "area":[]}
        for annotation in annotations:
            new_annotations["xmin"].append((annotation["x"]*xr_factor)/resize_to[1])
            new_annotations["xmax"].append(((annotation["x"]+annotation["width"])*xr_factor)/resize_to[1])
            new_annotations["ymin"].append((annotation["y"]*yr_factor)/resize_to[0])
            new_annotations["ymax"].append(((annotation["y"]+annotation["height"])*yr_factor)/resize_to[0])
            new_annotations["area"].append(
                (annotation["width"]*xr_factor*annotation["height"]*yr_factor) / \
                (resize_to[1]*resize_to[0])
            )
        return new_annotations
    else:
        raise NotImplementedError()

sub_train_df["effdet_annotations"] = sub_train_df.annotations.progress_apply(get_scaled_annotations)
sub_val_df["effdet_annotations"] = sub_val_df.annotations.progress_apply(get_scaled_annotations)

# DEMONSTRATE
plt.figure(figsize=(18,18))
plt.title("Original Image and Bounding Boxes", fontweight="bold")
plt.axis(False)
img = cv2.imread(sub_train_df.iloc[0]["img_path"])[..., ::-1].astype(np.uint8)
img = cv2.rectangle(img, 
                    (sub_train_df.iloc[0]["annotations"][0]["x"], sub_train_df.iloc[0]["annotations"][0]["y"]),
                    (sub_train_df.iloc[0]["annotations"][0]["x"]+sub_train_df.iloc[0]["annotations"][0]["width"], sub_train_df.iloc[0]["annotations"][0]["y"]+sub_train_df.iloc[0]["annotations"][0]["height"]),
                    (255, 0, 0), 2)
plt.imshow(img)
plt.show()

plt.figure(figsize=(18,18))
plt.title("Resized Image and Bounding Boxes", fontweight="bold")
plt.axis(False)
img = tf_load_image(sub_train_df.iloc[0].img_path).numpy()
img = cv2.rectangle(img, 
                    (int(sub_train_df.iloc[0]["effdet_annotations"]["xmin"][0]*INPUT_SHAPE[1]), 
                     int(sub_train_df.iloc[0]["effdet_annotations"]["ymin"][0]*INPUT_SHAPE[0])),
                    (int(sub_train_df.iloc[0]["effdet_annotations"]["xmax"][0]*INPUT_SHAPE[1]), 
                     int(sub_train_df.iloc[0]["effdet_annotations"]["ymax"][0]*INPUT_SHAPE[0])),
                    (255, 0, 0), 2)
plt.imshow(img)
plt.show()
```

```python
def model_inference(model, img, image_id=99999, score_thresh=0.1, cat_index={1: {'id': 1, "name":'COTS'}}, full_size=True, return_str=True, do_plot=False):
    def __predict(__imgs, __image_ids, return_as_numpy=True):
        cls_outputs, box_outputs = model(__imgs, training=False)
        nms_boxes_bs, nms_scores_bs, nms_classes_bs, _ = \
        postprocess.postprocess_global(config, cls_outputs, box_outputs)
        preds = postprocess.generate_detections_from_nms_output(
            nms_boxes_bs, nms_classes_bs, nms_scores_bs, _image_ids,
        )
        if return_as_numpy:
            return nms_classes_bs.numpy().astype(np.int32)[0], nms_boxes_bs.numpy()[0], nms_scores_bs.numpy()[0]
        else:
            return tf.cast(nms_classes_bs, tf.int32)[0], nms_boxes_bs[0], nms_scores_bs[0]
    
    __img = img.copy() # For display later
    _imgs = tf.expand_dims(adv_image_preprocess(tf.image.resize(img, INPUT_SHAPE[:-1])), axis=0)
    _image_ids = [image_id,]
    
    # Prediction Portion
    nms_classes_bs, nms_boxes_bs, nms_scores_bs = __predict(_imgs, _image_ids, return_as_numpy=True)
    
    # Change Back To Full Size
    if full_size:
        nms_boxes_bs[:, 0] = ((nms_boxes_bs[:, 0]/INPUT_SHAPE[0])*IMG_SHAPE[0])
        nms_boxes_bs[:, 1] = ((nms_boxes_bs[:, 1]/INPUT_SHAPE[1])*IMG_SHAPE[1])
        nms_boxes_bs[:, 2] = ((nms_boxes_bs[:, 2]/INPUT_SHAPE[0])*IMG_SHAPE[0])
        nms_boxes_bs[:, 3] = ((nms_boxes_bs[:, 3]/INPUT_SHAPE[1])*IMG_SHAPE[1])
    nms_boxes_bs = nms_boxes_bs.astype(np.int32)
    
    if do_plot:
        plt.figure(figsize=(20,20))
        __img = vis_utils.visualize_boxes_and_labels_on_image_array(
            __img,
            nms_boxes_bs,
            nms_classes_bs,
            nms_scores_bs,
            category_index=cat_index, 
            min_score_thresh=score_thresh,
            line_thickness=3
        )
        plt.imshow(__img)
        plt.show()

    return_str = ""
    for _cls, _box, _scr in zip(nms_classes_bs, nms_boxes_bs, nms_scores_bs):
        if _scr<score_thresh:
            continue
        else:
            return_str+=f" {_scr} {_box[1]} {_box[0]} {_box[3]-_box[1]} {_box[2]-_box[0]}"
    return return_str.strip()
```
