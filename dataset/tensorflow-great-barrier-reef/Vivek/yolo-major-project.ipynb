```python
!mkdir /kaggle/working/datasets/
!mkdir /kaggle/working/datasets/coco
!mkdir /kaggle/working/datasets/coco/images
!mkdir /kaggle/working/datasets/coco/labels

import pandas as pd

annotations_csv=pd.read_csv("../input/tensorflow-great-barrier-reef/train.csv")
final_anno_csv=pd.DataFrame({"image_id":[],"x":[],"y":[],"width":[],"height":[]})
for i in range(len(annotations_csv.index)):
    s=list(eval(annotations_csv["annotations"][i]))
    f=open(f'/kaggle/working/datasets/coco/labels/{annotations_csv["image_id"][i]}.txt',"a")
    for k in s:
        final_anno_csv.loc[len(final_anno_csv.index)]=[annotations_csv["image_id"][i],k["x"]/1280.0,k["y"]/720.0,k["width"]/1280.0,k["height"]/720.0]
        f.write(f'{0} {k["x"]/1280.0+k["width"]/2560.0} {k["y"]/720.0+k["height"]/1440.0} {k["width"]/1280.0} {k["height"]/720.0}'+"\n")

final_anno_csv["class"]=["star"]*len(final_anno_csv.index)
final_anno_csv["x_centre"]=final_anno_csv["x"]+final_anno_csv["width"]/2.0
final_anno_csv["y_centre"]=final_anno_csv["y"]+final_anno_csv["height"]/2.0
image_path=[]
for i in range(len(final_anno_csv.index)):
    image_path.append(f"/kaggle/input/tensorflow-great-barrier-reef/train_images/video_{final_anno_csv['image_id'][i][0]}/{final_anno_csv['image_id'][i][2:]}.jpg")

final_anno_csv["image_path"]=image_path



with open("/kaggle/working/yolov5/data/coco.yaml","r") as jkl:
    print(jkl.read())



data={"path":"/kaggle/working/datasets/coco","train":"images","val":"images","test":"images","nc":1,"names":["star"]}

import yaml

with open("/kaggle/working/yolov5/data/coco.yaml","w") as f:
    yaml.dump(data,f)
    
with open("/kaggle/working/yolov5/data/coco.yaml","r") as jkl:
    print(jkl.read())
```
