There are two starfish in this frame, let's add the bounding boxes

```python
annotations = train_df[train_df.annotations.map(len) > 100].iloc[0].annotations
annotations
```

```python
def get_bboxes(annotations):
    annots = ast.literal_eval(annotations)
    bboxes = []
    for annotation in annots:
        x, y = annotation['x'], annotation['y']
        w, h = annotation['width'], annotation['height']
        bboxes.append([x,y, x+w, y+h])
    return bboxes
```

```python
bboxes = get_bboxes(annotations)
bboxes
```

```python
def draw_bboxes(img, bboxes):
    draw = Draw(img)
    for bbox in bboxes:
        draw.rectangle(bbox, outline='red')
    return img
```

```python
draw_bboxes(img, bboxes)
```

Both starfish appropriately drawn in the image!
