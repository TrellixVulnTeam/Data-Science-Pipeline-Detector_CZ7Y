```python
#check if there are more than one annnotation for a frame (concl. there are more than one annotation for a frame)
annotation_len=[]
for i in range(0,len(train),5):
    ann=ast.literal_eval(train.iloc[i,5])
    annotation_len.append(len(ann))
print("First Annotation length",annotation_len[0])
print("Minimum annotation in one frame",min(annotation_len))
print("Maximum annotation in one frame",max(annotation_len))
print("mean of Annotation length",np.mean(annotation_len))
```

```python
#converts the annotation of string format  to list containing x,y,width,height in integer format
def get_bboxes(annotations):
    annotations=ast.literal_eval(annotations)
    if len(annotations)==0:
        return [[0,0,0,0]]
    boxes = pd.DataFrame(annotations, columns=['x', 'y', 'width', 'height']).astype(np.int32).values
    return boxes 
```

```python
#plotting annotation over image
def plot_annotation_on_img(video_id,video_frame,annotations):
    bbox=get_bboxes(annotations)
    img_url="../input/tensorflow-great-barrier-reef/train_images/video_"+str(video_id)+"/"+str(video_frame)+".jpg"
    img=plt.imread(img_url)
    for i in range(len(bbox)):        
        plt.gca().add_patch(Rectangle((bbox[i][0] ,bbox[i][1]), bbox[i][2], bbox[i][3],edgecolor='red',
                        facecolor='none',
                        lw=4))
    plt.rcParams["figure.figsize"] = (15,15)
    plt.axis("off")
    plt.imshow(img)
    print(video_id,video_frame,annotations)
```

```python
#given imnage number the annotation are draan on image
img_no=1345
vi=train.iloc[img_no,0]
vf=train.iloc[img_no,2]
ann=train.iloc[img_no,5]
plot_annotation_on_img(vi,vf,ann)
```
