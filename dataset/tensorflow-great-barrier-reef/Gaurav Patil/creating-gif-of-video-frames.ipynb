making gif 

```python
#making gif from video frames
def make_gif(frame_folder):
    frames=[]
    for i in range(0,450,5):
        frames.append(Image.open(f"{frame_folder}/{i}.jpg"))
    frames[0].save("video_0.gif", format="GIF", append_images=frames, save_all=True, duration=120, loop=0)
    
make_gif("../input/tensorflow-great-barrier-reef/train_images/video_0/")
#show the gif
HTML('<img src="video_0.gif" />')
```
