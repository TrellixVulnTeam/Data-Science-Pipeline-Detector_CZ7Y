> Counting the number of consecutive bases in a double strand

> Count of C in double strand

> Difference between C count and U count in 2 difference

> * 1−1ループはG-Gのペアが安定

> * 2−2ループはG-U配列が安定

> * ３塩基のヘアピンループは不安定
