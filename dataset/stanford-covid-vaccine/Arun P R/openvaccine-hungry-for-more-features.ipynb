## Feature 3

Rationale: Though this is a prediction, it gives some means of differentiating between unpaired bases (Note that all paired bases are of type S, if you haven't noticed that earlier).

## Feature 4

As provided in the dataset:Structure

Rationale: Include this only if you want redundancy in you training data. Why? All (,) in structure are of type S in Predicted Loop Type. So correlation between Predicted loop type S and structure is -1. I think you might have realised where I am going....else, do a little google search of you own why that matters.

## Feature 5

Derived Feature:Pair distance (from structure)

Rationale: match ( and ) in structure to get pair positions, get the absolute difference to get this feature. But be sure to checkout [Dangerous Features](https://www.kaggle.com/its7171/dangerous-features) before you decide to include this. Sorry for cheating again.

## Feature 6

Derived Feature:Pair distance from stem ends (from structure)

Rationale: Pairs in the middle of the stem are less reactive compared to those in the ends. How to compute this? It was a bit tricky. You match all pairs as in previous features. Now find pairs that are part of the same stem (beware that there can be multiple stems in a sequence). One trick I applied is using a queue to match (some public notebooks have done this) and a stem is complete when my queue is empty. Thus find the stems. Then take min of relative position from either end for each pair of the stem and you get the feature.

## Feature 7
