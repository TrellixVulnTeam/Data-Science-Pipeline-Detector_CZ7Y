```python
import pandas as pd
import numpy as np
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train=pd.read_csv("../input/stayalert/fordTrain.csv",index_col=1)
```

```python
test=pd.read_csv("../input/stayalert/fordTest.csv")
```

#counting missing values

```python
train.isna().sum()
```

```python
test.isna().sum()
```

```python
train.describe()
```

```python
'''
Since V7,V9  and P8 have all values zero, Trial ID not required.
we will drop these variables'''

train=train.drop(columns=["TrialID","V7","V9","P8"])
```

```python
train.info()
#check RAM uses
```

```python
#Reducing RAM uses

x=['P3','P6','E3','E4','E6','E7','E8','E9','E10','V3','V5','V6','V10','IsAlert']
for i in x:
    train[i]=train[i].astype('int16')

y=['P1','P2','P4','P5','P7','E1','E2','E5','E11','V1','V2','V4','V8','V11']
for i in y:
    train[i]=train[i].astype('float16')
```

```python
train.info()
```

```python
#model building

from sklearn.model_selection import train_test_split

x = train.drop(['IsAlert'], axis='columns')

y=train['IsAlert']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)


from sklearn.linear_model import LogisticRegression
```

```python
#initiate the model

logreg=LogisticRegression()

#fit the model with data

logreg.fit(x_train,y_train)

#predicting y_test

y_pred=logreg.predict(x_test)

#import metrics class

from sklearn import metrics

cnf_matrix=metrics.confusion_matrix(y_test,y_pred)

cnf_matrix

metrics.precision_score(y_test,y_pred)
metrics.recall_score(y_test,y_pred)
```

```python
#randomeforestclassifier

from sklearn.ensemble import RandomForestClassifier

rfc=RandomForestClassifier()
rfc.fit(x_train,y_train)


y_pred=rfc.predict(x_test)



```

```python
cnf_matrix1=metrics.confusion_matrix(y_pred,y_test)
print(y_test.value_counts())
print(cnf_matrix1)
```

```python
import seaborn as sns 
sns.heatmap(cnf_matrix1,annot=True,fmt="d")
```

```python
metrics.f1_score(y_test,y_pred)
```

```python
report=metrics.classification_report(y_test,y_pred)
print(report)
```

```python
from sklearn import tree


clf=tree.DecisionTreeClassifier()
clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)
cnf_matrix1=metrics.confusion_matrix(y_pred,y_test)
print(metrics.classification_report(y_test,y_pred))
```
