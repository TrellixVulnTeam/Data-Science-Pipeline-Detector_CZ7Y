```python
#Importing Libraries
import pandas as pd
import numpy as np
import os
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import tensorflow as tf
```

```python
train=pd.read_csv("/kaggle/input/stayalert/fordTrain.csv")
```

```python
#counting missing values
train.isna().sum()
```

```python
pd.set_option("display.max_columns",500)
train.describe()
```

```python
plt.hist(x=train["P6"])
train["IsAlert"].value_counts()
```

We are deleting rows which have P5 values greater than 0.6 because they are outliers
We will delete values which is above 0.35 which is 2% (12088)


```python
train=train[train["P6"]<=30000]
train=train.drop(columns=["TrialID","V7","V9","P8"])
x=['P3','P6','E3','E4','E6','E7','E8','E9','E10','V3','V5','V6','V10','IsAlert']
for i in x:
    train[i]=train[i].astype('int16')
y=['P1','P2','P4','P5','P7','E1','E2','E5','E11','V1','V2','V4','V8','V11']
for i in y:
    train[i]=train[i].astype('float16')
```

```python
from sklearn.model_selection import train_test_split

x = train.drop(['IsAlert'], axis='columns')



y=train['IsAlert']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
import tensorflow as tf
```

```python
with tf.device('/device:GPU:0'):
    feature_selection_model = RandomForestClassifier()
    feature_selection_model.fit(x_train,y_train)
```

```python
with tf.device('/device:GPU:0'):
    print(feature_selection_model.score(x_test,y_test))
```

```python
with tf.device('/device:GPU:0'):
    from sklearn.svm import SVC
    model=SVC()
    model.fit(x_train,y_train)
```
