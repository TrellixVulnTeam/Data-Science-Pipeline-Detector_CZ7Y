Importing various libraries needed for the EDA

```python
import pandas as pd                     
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
import missingno
```

```python
df = pd.read_csv('/kaggle/input/tabular-playground-series-apr-2021/train.csv')
df.head()
```

<h1 style='background:#1E30E9; border:0; color:white'><center>Explaining the Columns</center></h1> 


1> **PassengerID** - Unique id assigned to every passenger


2> **Survived** - Target variable (0 - if the passenger died , 1 - if the passenger survived)

3> **PClass** - Passenger class (1- Upper Class , 2- Middle Class , 3- Lower Class) - Fun Fact > Class 3 ticket costed around 15 to 40 dollars.

4> **Name** - Passenger Name

5> **Sex** - Gender of the passenger

6> **Age** - Age in years

7> **SibSp** - Number of the passenger's siblings and spouse

8> **Parch** - Number of parents and children for the passenger

9> **Ticket** - The Ticket Number

10> **Fare** - Ticket cost for the passenger (need to confirm my 'Fun Fact')

11> **Cabin** - Passenger's Cabin Number

12> **Embarked** - The port where the Passenger set sail (C, Q and S)

<h1 style='background:#1E30E9; border:0; color:white'><center>EDA</center></h1> 

```python
missingno.bar(df,figsize= (9,6),color= 'red' )
```

Seems like we are missing data in Age , Ticket , Embarked and Cabin columns

Survivor Distribution

```python
val  = df['Survived'].value_counts()
px.pie(names = val.index, values = val.values ,title = 'Survived v/s Not Survived', hole = 0.2)
```

57.2% People onboard died 


42.8% Survived

Comparing Sex and Survival rate to know if or not class had any affect on the passenger's survival

```python
grp = df.groupby('Sex').count()['Survived']
px.bar(x = grp.index,y = grp.values, title = 'Surivial grouped by sex',height=700,width =600,color=grp.values )
```
