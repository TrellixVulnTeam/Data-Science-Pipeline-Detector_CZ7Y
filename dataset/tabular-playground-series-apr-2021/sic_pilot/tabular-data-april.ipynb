```python
# Cabin

train['Cabin'] = train['Cabin'].fillna('NA')


test['Cabin'] = test['Cabin'].fillna('NA')

#Embarked
train['Embarked'] = train['Embarked'].fillna('NA')

test['Embarked'] = test['Embarked'].fillna('NA')

# Fare
train['Fare'] = train['Fare'].fillna(np.mean(train['Fare']))

test['Fare'] = test['Fare'].fillna(np.mean(test['Fare']))

# Ticket
train['TicketCode'] = train['Ticket'].str.replace('[^\w\s]','')
train['TicketCode'] = train['TicketCode'].str.replace(' ','')
train['TicketCode'] = train['TicketCode'].fillna('NA')
train['TicketCode'] = train['TicketCode'].replace('(\d)', '', regex=True)
train['TicketNumber'] = train['Ticket'].str.extract('(\d+)')
train['TicketNumber'] = train['TicketNumber'].astype(float)
train['TicketNumber'] = train['TicketNumber'].fillna(0)

test['TicketCode'] = test['Ticket'].str.replace('[^\w\s]','')
test['TicketCode'] = test['TicketCode'].str.replace(' ','')
test['TicketCode'] = test['TicketCode'].fillna('NA')
test['TicketCode'] = test['TicketCode'].replace('(\d)', '', regex=True)
test['TicketNumber'] = test['Ticket'].str.extract('(\d+)')
test['TicketNumber'] = test['TicketNumber'].astype(float)
test['TicketNumber'] = test['TicketNumber'].fillna(0)

# Age 
train['AgeBin']=pd.cut(train['Age'],[-np.inf, 20, 40, 60, 80, np.inf], right=False)

test['AgeBin']=pd.cut(test['Age'],[-np.inf, 20, 40, 60, 80, np.inf], right=False)
```
