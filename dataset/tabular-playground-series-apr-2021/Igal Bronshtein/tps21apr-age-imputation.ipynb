```python
import pandas as pd
from sklearn.metrics import mean_absolute_error

df = pd.read_csv('../input/tabular-playground-series-apr-2021/train.csv')
```

```python
full_rows = df.dropna().copy()
full_rows["Age01_mean"] = df.Age.mean()
```

```python
mae_age01_mean = mean_absolute_error(full_rows.Age, full_rows["Age01_mean"])
precision = 3
round(mae_age01_mean, precision)
```

```python
male_mean_age = df[df.Sex == 'male'].Age.mean()
male_mean_age

female_mean_age = df[df.Sex == 'female'].Age.mean()
female_mean_age

full_rows["Age02_sex"] = full_rows.Sex.apply(lambda x: female_mean_age if (x == 'female') else male_mean_age)
```

```python
mae_age02_sex = mean_absolute_error(full_rows.Age, full_rows["Age02_sex"])
round(mae_age02_sex, precision)
```

```python
first_class_mean_age = df[df.Pclass == 1].Age.mean()
second_class_mean_age = df[df.Pclass == 2].Age.mean()
third_class_mean_age = df[df.Pclass == 3].Age.mean()
```

```python
def age_pclass(x):
    return { 1 : first_class_mean_age, 2: second_class_mean_age, 3: third_class_mean_age}[x]

full_rows["Age03_pclass"] = full_rows.Pclass.apply(lambda x: age_pclass(x))
```

```python
mae_age03_pclass =  mean_absolute_error(full_rows.Age, full_rows["Age03_pclass"])
round(mae_age03_pclass, precision)
```

```python
sibSp0_mean_age = df[df.SibSp == 0].Age.mean()
sibSp1_mean_age = df[df.SibSp == 1].Age.mean()
sibSp2_mean_age = df[df.SibSp == 2].Age.mean()
sibSp3_mean_age = df[df.SibSp == 3].Age.mean()
sibSp4_mean_age = df[df.SibSp == 4].Age.mean()
sibSp5_mean_age = df[df.SibSp == 5].Age.mean()
sibSp8_mean_age = df[df.SibSp == 8].Age.mean()
```

```python
def age_sibsp(x):
    return { 0 : sibSp0_mean_age, 1: sibSp1_mean_age, 2: sibSp2_mean_age, 
           3: sibSp3_mean_age, 4: sibSp4_mean_age, 5: sibSp5_mean_age, 8: sibSp8_mean_age}[x]

full_rows["Age04_sibsp"] = full_rows.SibSp.apply(lambda x: age_sibsp(x))
```

```python
mae_age04_sibsp = mean_absolute_error(full_rows.Age, full_rows["Age04_sibsp"])
round(mae_age04_sibsp, precision)
```

```python
parch0_mean_age = df[df.Parch == 0].Age.mean()
parch1_mean_age = df[df.Parch == 1].Age.mean()
parch2_mean_age = df[df.Parch == 2].Age.mean()
parch3_mean_age = df[df.Parch == 3].Age.mean()
parch4_mean_age = df[df.Parch == 4].Age.mean()
parch5_mean_age = df[df.Parch == 5].Age.mean()
parch6_mean_age = df[df.Parch == 6].Age.mean()
parch9_mean_age = df[df.Parch == 9].Age.mean()
```

```python
def age_parch(x):
    return { 0 : parch0_mean_age, 1: parch1_mean_age, 2: parch2_mean_age, 
           3: parch3_mean_age, 4: parch4_mean_age, 5: parch5_mean_age, 6: parch6_mean_age, 9: parch9_mean_age}[x]

full_rows["Age05_parch"] = full_rows.Parch.apply(lambda x: age_parch(x))
```

```python
mae_age05_parch = mean_absolute_error(full_rows.Age, full_rows["Age05_parch"])
round(mae_age05_parch, precision)
```

```python
class1_c_mean_age = df[(df.Pclass == 1) & (df.Embarked == 'C')].Age.mean()
class1_q_mean_age = df[(df.Pclass == 1) & (df.Embarked == 'Q')].Age.mean()
class1_s_mean_age = df[(df.Pclass == 1) & (df.Embarked == 'S')].Age.mean()

class2_c_mean_age = df[(df.Pclass == 2) & (df.Embarked == 'C')].Age.mean()
class2_q_mean_age = df[(df.Pclass == 2) & (df.Embarked == 'Q')].Age.mean()
class2_s_mean_age = df[(df.Pclass == 2) & (df.Embarked == 'S')].Age.mean()

class3_c_mean_age = df[(df.Pclass == 3) & (df.Embarked == 'C')].Age.mean()
class3_q_mean_age = df[(df.Pclass == 3) & (df.Embarked == 'Q')].Age.mean()
class3_s_mean_age = df[(df.Pclass == 3) & (df.Embarked == 'S')].Age.mean()
```

```python
full_rows["ClassEmbarkedString"] = full_rows["Pclass"].astype('str') + full_rows["Embarked"] 
```

```python
def class_embarked_tree(x):
    return { '1C' : class1_c_mean_age, '1Q': class1_q_mean_age, '1S': class1_s_mean_age,
           '2C' : class2_c_mean_age, '2Q': class2_q_mean_age, '2S': class2_s_mean_age,
           '3C' : class3_c_mean_age, '3Q': class3_q_mean_age, '3S': class3_s_mean_age}[x]

full_rows["Age09_ClassEmbarkedTree"] = full_rows.ClassEmbarkedString.apply(lambda x: class_embarked_tree(x))
```

```python
mae_age09_class_embarked_tree = mean_absolute_error(full_rows.Age, full_rows["Age09_ClassEmbarkedTree"])
round(mae_age09_class_embarked_tree, precision)
```
