```python
df1 = df.copy()
test1 = test.copy()
```

initially i am droping cabing and clearing all empty values , later on we will replace it with different values that best fits

```python
df1 = df1.drop('Cabin',axis=1)
```

```python
df1 = df1.dropna()
```

```python
test1 = test1.drop('Cabin',axis=1)
test1 = test1.dropna()
```

```python
sns.heatmap(df1.isnull())
```

```python
df1.head(2)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df1['Sex'] = le.fit_transform(df1['Sex']) 
test1[['Sex']] = le.fit_transform(test1['Sex']) 
```

```python
df1['Embarked'] = le.fit_transform(df1['Embarked'])
test1['Embarked'] = le.fit_transform(test1['Embarked'])
```

```python
df1.head(1)
```

```python
test1.head()
```

```python
df1
```

```python
x = df1[['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']]
y = df1['Survived']
#test1 = test1.drop('PassengerId',axis=1)
#test1 = test1.drop('Name',axis=1)
```

```python
test1 = test1.drop('Ticket',axis=1)
test1.head(5)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1,stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
log.fit(x_train,y_train)
log.score(x_test,y_test)
```
