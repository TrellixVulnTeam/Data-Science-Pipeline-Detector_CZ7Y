```python
train = pd.read_csv('../input/tabular-playground-series-apr-2021/train.csv')
test = pd.read_csv('../input/tabular-playground-series-apr-2021/test.csv')
```

```python
train.head(2)
```

<h2>Import the module in your script <h2>

```python
from dfencoding import utilities
```

<h2>Instantiation :<h2>

 

More information with help(utilities.dfencoding)

```python
dfe = utilities.dfencoding(train,"Survived",test,missing_value='Y')
```

<h2> Simplest method : encode() <h2>
    label encoding (for categories) + Minmax encoding (for all)

```python
dfe.encode()
```

```python
dfe.decode()
```

<h2> To retrieve the dataframe encoded <h2>

```python
dfe.encode()
# you can get the attribute self.train, self.test, self.data (train+test) :
train_encoded = dfe.train
train_encoded.head(3)
```

```python
dfe.decode()
```

<h2> Label encoding <h2>

```python
dfe.labelencode()
# reminder : you can change the limit of number of categories per column for encoding during instantiation
```

```python
dfe.labeldecode()
```

<h2> Target encoding <h2>

```python
dfe.targetencode()
```

```python
dfe.targetdecode()
```

<h2> Minmax encoding <h2>

```python
dfe.minmaxencode()
```

```python
dfe.minmaxdecode()
```
