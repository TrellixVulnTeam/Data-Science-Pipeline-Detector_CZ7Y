## SweetViz

Sweetviz is an open-source Python library that generates beautiful, high-density visualizations to kickstart EDA (Exploratory Data Analysis) with just two lines of code. Output is a fully self-contained HTML application.

```python
!pip install sweetviz
```

```python
import pandas as pd

train = pd.read_csv('../input/tabular-playground-series-apr-2021/train.csv')
test = pd.read_csv('../input/tabular-playground-series-apr-2021/test.csv')
```

Comparing Train and Test Datasets and using *Survived* as the Target.

```python
import sweetviz as sv

comparison=sv.compare([train,'Train'],[test,'Test'], "Survived") 
comparison.show_html('Comparison.html')
```

*Comparison.html* can be downloaded and viewed from the Kaggle Output directory

# Autoviz

Automatically Visualize any dataset, any size with a single line of code.
AutoViz performs automatic visualization of any dataset with one line. Give any input file (CSV, txt or json) and AutoViz will visualize it.

```python
!pip install autoviz
!pip install xlrd
```

```python
from autoviz.AutoViz_Class import AutoViz_Class
AV = AutoViz_Class()
```

```python
sep = ','
dft = AV.AutoViz(filename="",sep=sep, depVar='Survived', dfte=train, header=0, verbose=1, 
                 lowess=False, chart_format='svg', max_rows_analyzed=150000, max_cols_analyzed=30)
```
