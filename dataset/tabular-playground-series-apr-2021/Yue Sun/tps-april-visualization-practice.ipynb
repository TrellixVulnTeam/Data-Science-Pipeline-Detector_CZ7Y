```python
import plotly as py
import plotly.graph_objs as go
import plotly.tools as tls
from plotly.offline import iplot
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
sample = pd.read_csv("/kaggle/input/tabular-playground-series-apr-2021/sample_submission.csv")
train = pd.read_csv("/kaggle/input/tabular-playground-series-apr-2021/train.csv")
test = pd.read_csv("/kaggle/input/tabular-playground-series-apr-2021/test.csv")
train
```

```python
iplot(go.Figure(data = go.Bar(
     x = train['Survived'].value_counts().index ,
     y = train['Survived'].value_counts().values,
     marker=dict(
         color= train['Survived'].value_counts().values,
         colorscale = 'Viridis',
         reversescale = True),
     name = "Imbalance"), layout= dict(
     title = "Data imbalance - Survived",
     width = 500, height = 500,
     xaxis = go.layout.XAxis(
     automargin=True),
     yaxis = dict(
         showgrid = False,
         showline = False,
         showticklabels = True))))
```

```python
sns.set_context({"figure.figsize" : (15,5)})
sns.kdeplot(x= train['Age'], shade=True, color='olive', alpha = 0.6, zorder= 3,linewidth=5,edgecolor='black')
```

```python
sns.set_context({"figure.figsize" : (20,5)})
sns.histplot(x= train['Fare'], color='olive', alpha=1,zorder=2,linewidth=1,edgecolor='black',shrink=0.5)
```

```python
fig=plt.figure(figsize=(15,7),facecolor='white')
gs=fig.add_gridspec(1,1)
ax0=fig.add_subplot(gs[0,0])
ax0.text(-0,1.4,"Pclass Distribution",fontsize=18,fontweight='bold', fontfamily='monospace',ha='center')
colormap1=['firebrick','LightCoral','Salmon']
ax0.pie(train['Pclass'].value_counts(),  explode = [0.1, 0, 0], labels = train['Pclass'].value_counts().index, autopct='%1.1f%%',
        shadow=True, startangle=90,colors=colormap1)
ax0.axis('equal')
plt.tight_layout()
plt.show()
```
