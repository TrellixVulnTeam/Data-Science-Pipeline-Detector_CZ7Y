```python
from supervised.automl import AutoML
import pandas as pd
df_train = pd.read_csv('../input/tabular-playground-series-apr-2021/train.csv')
df_test = pd.read_csv('../input/tabular-playground-series-apr-2021/test.csv')
```

```python
X_train = df_train[df_train.columns[2:]]
y_train = df_train[df_train.columns[1]]
X_test = df_test[df_train.columns[2:]]
```

```python
automl = AutoML(mode='Compete', total_time_limit= total_time * 3600, hill_climbing_steps=3, top_models_to_improve=5, n_jobs=-1, random_state=990121)
automl.fit(X_train, y_train)
preds = automl.predict(X_test)
submission = pd.DataFrame({'PassengerId':df_test.PassengerId, 'Survived': preds})
submission.to_csv('1_submission.csv', index=False)
```
