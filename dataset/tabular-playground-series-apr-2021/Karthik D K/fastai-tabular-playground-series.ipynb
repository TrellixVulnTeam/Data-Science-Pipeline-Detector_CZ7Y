```python
from fastai.tabular.all import *
```

```python
df = pd.read_csv("../input/tabular-playground-series-apr-2021/train.csv", low_memory=False)
df.head()
```

```python
df.info()
```

```python
dft = pd.read_csv("../input/tabular-playground-series-apr-2021/test.csv", low_memory=False)
dft.head()
```

```python
dft.info()
```

```python
df.drop(["PassengerId", "Name", "Ticket", "Cabin"], axis=1, inplace=True)
dft.drop(["PassengerId", "Name", "Ticket", "Cabin"], axis=1, inplace=True)
df.shape, dft.shape
```

```python
to = TabularPandas(df, procs=[Categorify, FillMissing, Normalize],
                   cat_names = ['Pclass', 'Sex', 'SibSp', 'Parch', 'Embarked'],
                   cont_names = ['Age', 'Fare'],
                   y_names='Survived',
                   y_block = CategoryBlock(),
                   splits=RandomSplitter(valid_pct=0.2)(range_of(df)))
to.xs.iloc[:2]
```

```python
dls = to.dataloaders(bs=64)
dls.show_batch()
```

```python
learn = tabular_learner(dls, metrics=accuracy)
learn.fit_one_cycle(1)
```

```python
learn.summary()
```

```python
learn.show_results()
```

```python
row, clas, probs = learn.predict(df.iloc[0])
row.show()
```

```python
clas, probs
```

```python
test_df = dft.copy()
dl = learn.dls.test_dl(test_df)
```

```python
probs, _ = learn.get_preds(dl=dl)
```

```python
preds = torch.argmax(probs, dim=1)
```

```python
preds[0:5], preds.shape
```

```python
df_test = pd.read_csv("../input/tabular-playground-series-apr-2021/test.csv")
output = pd.DataFrame({'PassengerId': df_test.PassengerId, 'Survived': preds})
output.to_csv('fastai_R00.csv', index=False)
print("Your submission was successfully saved!")
```
