```python
from sklearn.metrics import classification_report
```

```python
from sklearn.preprocessing import StandardScaler, Normalizer
```

```python
sts = StandardScaler()
nrm = Normalizer()
```

```python
x_new = nrm.fit_transform(x)
x_new = sts.fit_transform(x_new)
```

```python
x_train, x_val, y_train, y_val = train_test_split(x_new, y, test_size=0.01, random_state=42)
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
mlp = MLPClassifier(hidden_layer_sizes=(20, 100, 150, 200),random_state=42,early_stopping=True)
```

```python
mlp.fit(x_train, y_train)
```

```python
print(classification_report(y_val, mlp.predict(x_val)))
```
