# Load submissions dataset

The dataset I use here consists of submissions from [my kernel](https://www.kaggle.com/alexryzhkov/tps-april-21-lightautoml-starter) and MLjar [kernel](https://www.kaggle.com/mt77pp/mljar-automl-tps-apr-21?scriptVersionId=58426457) and [kernel](https://www.kaggle.com/mt77pp/mljar-automl-tps-apr-2021-manual-features?scriptVersionId=58538953)

```python
data = pd.read_csv('../input/submitstpsautomlkernels/submits_from_kernels.csv')
data
```

Dropping 4 bad score submits:

```python
data.drop(columns = ['submit_079058', 'submit_079268', 'submit_079397', 'submit_079688'], inplace = True)
```

Calculate votes:

```python
data[[col for col in data.columns if col.startswith('submit_')]].sum(axis = 1).value_counts()
```

Ok, the majority of objects have all ones or all zeros - this is cool. Let's make voting for another objects:
- 2 goes to 1 in submission
- 1 goes to 0 in submission

```python
data['Survived'] = (data[[col for col in data.columns if col.startswith('submit_')]].sum(axis = 1) >= 2).astype(int)
data
```

```python
data[['PassengerId', 'Survived']].to_csv('voting_submission_from_3_best.csv', index = False)
```
