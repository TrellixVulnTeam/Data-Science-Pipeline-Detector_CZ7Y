```python
titanic_test['Age'] = titanic_test['Age'].fillna((titanic_test['Age'].mean()))
titanic_test['Fare'] = titanic_test['Fare'].fillna((titanic_test['Fare'].mean()))
```

```python
sex=pd.get_dummies(titanic_test['Sex'],drop_first=True)
embark=pd.get_dummies(titanic_test['Embarked'],drop_first=True)
pcl=pd.get_dummies(titanic_test['Pclass'],drop_first=True)
titanic_test=pd.concat([titanic_test,sex,embark,pcl],axis=1)
titanic_test.drop(['Sex','Embarked','Pclass'],axis=1,inplace=True)
```

### Training Model

```python
x_train = titanic_data.drop(['Survived'], axis=1)
y_train = titanic_data['Survived']
x_test=titanic_test
```

```python
from sklearn.linear_model import LogisticRegression
logmodel=LogisticRegression(max_iter=300)
logmodel.fit(x_train,y_train)
predictions=logmodel.predict(x_test)
df["Survived"]=predictions
df.head()
```
