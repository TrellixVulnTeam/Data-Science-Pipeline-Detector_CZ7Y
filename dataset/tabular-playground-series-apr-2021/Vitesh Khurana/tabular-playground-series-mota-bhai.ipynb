```python
pclass1_mean=train_data[train_data['Pclass']==1]['Fare'].mean()
pclass2_mean=train_data[train_data['Pclass']==2]['Fare'].mean()
pclass3_mean=train_data[train_data['Pclass']==3]['Fare'].mean()
pclass_mean={1:pclass1_mean,2:pclass2_mean,3:pclass3_mean}
```

```python
def fill_missing(val,pclass):
    if pd.isnull(val):
        return(pclass_mean[pclass])
    else:
        return(val)
```

```python
train_data['Fare']=train_data.apply(lambda x:fill_missing(x['Fare'],x['Pclass']),axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler().fit(train_data[['Age','Fare']])
train_data[['Age','Fare']]=scaler.transform(train_data[['Age','Fare']])
```

```python
X=train_data.drop(['Survived','Name','Cabin','Ticket'],axis=1)
```

```python
y=train_data['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=18,test_size=0.25)
```
