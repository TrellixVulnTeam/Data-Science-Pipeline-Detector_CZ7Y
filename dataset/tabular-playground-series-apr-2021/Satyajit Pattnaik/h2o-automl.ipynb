```python
import numpy as np
import pandas as pd
import h2o
import matplotlib.pyplot as plt
import seaborn as sns
from h2o.automl import H2OAutoML
```

```python
h2o.init()
```

```python
train_df = h2o.import_file('../input/tabular-playground-series-apr-2021/train.csv',destination_frame='train_df')
```

```python
test_df = h2o.import_file('../input/tabular-playground-series-apr-2021/test.csv')
```

```python
train_df.describe()
```

```python
train_df = train_df.drop('PassengerId')
```

```python
predictor_cols = train_df.columns[1:]
```

```python
train, valid, test = train_df.split_frame(ratios=[0.6,0.2], seed=1111)
response = "Survived"
train[response] = train[response].asfactor()
valid[response] = valid[response].asfactor()
test[response] = test[response].asfactor()
print("Number of rows in train, valid and test set : ", train.shape[0], valid.shape[0], test.shape[0])
```

```python
aml = H2OAutoML(max_models=7, seed=1)
aml.train(x=predictor_cols, y=response, training_frame=train)
```

```python
aml.leaderboard
```

```python
test_df = test_df.drop('PassengerId')
```

```python
pred = aml.predict(test_df)
pred.head()
```

```python
p1 = pred['predict']
```

```python
p1_list = h2o.as_list(p1)
```

```python
submission = pd.read_csv('../input/tabular-playground-series-apr-2021/sample_submission.csv')
```

```python
submission
```

```python
submission['Survived'] = p1_list['predict'].values
```

```python
submission.to_csv('Submission_AutoML.csv',index=False)
```
