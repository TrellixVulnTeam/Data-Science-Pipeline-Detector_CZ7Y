# Import libraries

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

# Get outputs csv:

```python
output1 = pd.read_csv("../input/tps-apr-2021-submissions-democracy/voting_submission_from_3_best.csv")
output2 = pd.read_csv("../input/tps-apr-2021-lightgbm-cv/submission.csv")
output3 = pd.read_csv("../input/tps-apr-2021-single-lgbmregressor/submission.csv")
output4 = pd.read_csv("../input/tabular-playground-series-april-xgboost-automl-lr/DT.csv")
output5 = pd.read_csv("../input/tabular-playground-series-april-xgboost-automl-lr/LR.csv")
output6 = pd.read_csv("../input/tabular-playground-series-april-xgboost-automl-lr/XGB.csv")
#output7 = pd.read_csv("../input/n3-tps-april-21-lightautoml-starter/LightAutoML_compose_version_23.csv")
```

```python
total_result = (
    output1.Survived
    + output2.Survived
    + output3.Survived
    + output4.Survived
    + output5.Survived
    + output6.Survived
) / 6
output1.Survived = np.where(total_result > 0.5, 1, 0)
```

```python
output1.to_csv("lazy_submission.csv", index=0)
```
