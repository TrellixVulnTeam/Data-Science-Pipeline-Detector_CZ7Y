# 1. Introduction

**Kaggle competitions** are incredibly fun and rewarding, but they can also be intimidating for people who are relatively new in their data science journey. In the past, Kaggle have launched many Playground competitions that are more approachable than Featured competition, and thus more beginner-friendly.

The goal of these competitions is to provide a fun, but less challenging, tabular dataset. These competitions will be great for people looking for something in between the Titanic Getting Started competition and a Featured competition.

The dataset is used for this competition is synthetic but based on a real dataset (in this case, the actual **Titanic data**!) and generated using a CTGAN. The statistical properties of this dataset are very similar to the original Titanic dataset, but there's no way to "cheat" by using public labels for predictions.

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_palette("coolwarm")

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

# You can write up to 20GB to the current directory (/kaggle/working/) that gets preserved as output when you create a version using "Save & Run All" 
# You can also write temporary files to /kaggle/temp/, but they won't be saved outside of the current session
```

```python
# Loading the datasets
path = "/kaggle/input/tabular-playground-series-apr-2021/"
train = pd.read_csv(path + 'train.csv', index_col = "PassengerId")
test = pd.read_csv(path + 'test.csv', index_col = "PassengerId")
submission = pd.read_csv(path + 'sample_submission.csv')
```

# 2. Exploratory Data Analysis

Imagine that you want to buy a new mobile phone. We all know that there are variety of good phones out there in the Market. There's Apple, Samsung, Oppo, OnePlus, Google, etc. How are you going to decide which phone to buy now? You'll probably go to Youtube and watch MKBHD or Linus Tech Tips or MrWhoseTheBoss videos and will narrow down your choices. Is that it? If I was at your position, I would have compared the specifications of the phones too and compare them out so that I can narrow down my choices on a more deeper level now and I think that you would also do the same. 

Now this **investigation** that you or me are doing before buying a new phone is what in today's technological world or Data Science world known as "Exploratory Data Analysis". 

**Exploratory Data Analysis** is an approach to analyze any data set to summarize its main features or characteristics. We use statistical graphs and other data visulization models most of the time here.

### 2.1.1. Checking Null/Missing Values

We can check **the null values** in the dataset in 2 particular ways:
1. By using `is_na()` function, or
2. By using Seaborn and visualize it.

I mostly prefer to **visualize it first** and then see the numerical values.

```python
plt.figure(figsize=(10,4), facecolor="#f7e4e4")
ax = plt.axes()
ax.set_facecolor("#f7e4e4")
sns.heatmap(train.isnull(), cbar=False, yticklabels=False, cmap="coolwarm")
plt.title("Figure 1: HeatMap to check Null values in Train Dataset", fontsize=16)
plt.show()
```

### 2.2.3. Survived

This section explores the `Survived` feature of the Training dataset and findings of it with other features too.

**Observations:**

* `Less than 50%` of the passengers were able to `survive`.
* Among those who were able to survive, `females` were more prone to Survive than Males.
* Most of the passengers with `age 15-40` were not able to survive.
* `Older passengers` survival probablity was higher than the others. 
* Those having Ticket `class 1` had the highest chance to survived. Then came the Ticket class 2 and then class 3rd one's.
* Survival rate of the passengers `without siblings / spouses` to survive was the highest. That includes those too `who didn't brought parents or children` with them.
* Passengers who bought a `low fare` ticket were able to survive the most. [Doesn't make any sense though!]
* It's hard to predict in a strong sense when it comes to `Cabin` feature as it includes most of the `Null` values or `Missing` values.
* Passengers from `Southhampton` had the better probablity of surviving.
