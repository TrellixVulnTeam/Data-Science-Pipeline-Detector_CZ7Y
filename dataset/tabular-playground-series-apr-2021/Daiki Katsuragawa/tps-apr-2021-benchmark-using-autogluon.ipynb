```python
import pandas as pd
from autogluon.tabular import TabularDataset, TabularPredictor

train = TabularDataset('../input/tabular-playground-series-apr-2021/train.csv')
test = TabularDataset('../input/tabular-playground-series-apr-2021/test.csv')

label = 'Survived'
time_limit = 600
predictor = TabularPredictor(label=label,learner_kwargs={'ignored_columns': ['PassengerId']}).fit(train_data=train,time_limit=time_limit)

submission = pd.read_csv('../input/tabular-playground-series-apr-2021/sample_submission.csv')
submission[label] = predictor.predict(test)
submission.to_csv('submission.csv', index=False)
submission.head()
```
