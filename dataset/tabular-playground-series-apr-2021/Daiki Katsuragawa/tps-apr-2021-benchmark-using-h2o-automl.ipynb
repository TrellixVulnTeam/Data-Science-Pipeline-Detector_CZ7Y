```python
import pandas as pd
import h2o
from h2o.automl import H2OAutoML

h2o.init()
train = h2o.import_file('../input/tabular-playground-series-apr-2021/train.csv')
test = h2o.import_file('../input/tabular-playground-series-apr-2021/test.csv')

label ='Survived'
columns = train.columns
columns.remove(label)
aml = H2OAutoML()
train[label] = train[label].asfactor()
aml.train(x=columns,y=label,training_frame=train)

model = aml.leader
preds = model.predict(test)
preds_df = h2o.as_list(preds)

submission = pd.read_csv('../input/tabular-playground-series-apr-2021/sample_submission.csv')
submission[label] = preds_df['predict']
submission.to_csv('submission.csv', index=False)
submission.head()
```
