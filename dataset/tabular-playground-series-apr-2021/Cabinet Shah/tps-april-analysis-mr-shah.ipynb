# Replacing Null value with median 

```python
def rep_na(series):
    return series.fillna(series.median())
```

```python
test_data.Age = test_data['Age'].transform(rep_na)
```

**We can easily observe that null values in Age attribute has been replaced with median value **

```python
test_data.isnull().sum()
```

```python
missing_per_test = (test_data[test_data.columns].isnull().sum()/test_data.shape[0])*100
missing_per_test
```
