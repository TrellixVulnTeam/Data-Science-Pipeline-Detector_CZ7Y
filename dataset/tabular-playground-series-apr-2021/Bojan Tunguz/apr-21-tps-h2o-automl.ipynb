```python
# Starting H2O

import h2o
print(h2o.__version__)
from h2o.automl import H2OAutoML

h2o.init(max_mem_size='16G')
```

```python
%%time
train = h2o.import_file("../input/tps-april-2021-feature-eng-only/train.csv")
test = h2o.import_file("../input/tps-april-2021-feature-eng-only/test.csv")
```

```python
train.head()
```

```python
len(test.columns)
```

```python
test.columns
```

```python
x = test.columns
y = 'Survived'
train[y] = train[y].asfactor()
```

```python
# Run AutoML for 20 base models (limited to 1 hour max runtime by default)
aml = H2OAutoML(max_models=10000, seed=47, max_runtime_secs=1800)
aml.train(x=x, y=y, training_frame=train)
```

```python
# View the AutoML Leaderboard
lb = aml.leaderboard
lb.head(rows=lb.nrows)  # Print all rows instead of default (10 rows)
```

```python
# The leader model is stored here
aml.leader
```

```python
h2o.remove(train)
```

```python
preds = aml.predict(test)
preds['p1'].as_data_frame().values.flatten()
```
