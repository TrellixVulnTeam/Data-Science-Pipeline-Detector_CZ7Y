```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, accuracy_score
```

```python
sample_submission = pd.read_csv('/kaggle/input/tabular-playground-series-apr-2021/sample_submission.csv')
train = pd.read_csv('/kaggle/input/tabular-playground-series-apr-2021/train.csv')
test = pd.read_csv('/kaggle/input/tabular-playground-series-apr-2021/test.csv')
```

```python
train['Sex'] = 1*(train['Sex'] == 'male')
test['Sex'] = 1*(test['Sex'] == 'male')
```

```python
test['Age'].fillna((train['Age'].mean()), inplace=True)
train['Age'].fillna((train['Age'].mean()), inplace=True)
test['Fare'].fillna((train['Fare'].mean()), inplace=True)
train['Fare'].fillna((train['Fare'].mean()), inplace=True)
test['SibSp'].fillna((train['SibSp'].mean()), inplace=True)
train['SibSp'].fillna((train['SibSp'].mean()), inplace=True)
test['Parch'].fillna((train['Parch'].mean()), inplace=True)
train['Parch'].fillna((train['Parch'].mean()), inplace=True)
```

```python
columns = ['Sex', 'Age', 'Fare', 'SibSp', 'Parch']
target = train['Survived'].values
```

```python
lr = LogisticRegression(C=0.5)
lr.fit(train[columns], target)
```

```python
roc_auc_score(target, lr.predict_proba(train[columns])[:,1])
```

```python
accuracy_score(target, lr.predict(train[columns]))
```

```python
sample_submission['Survived'] = lr.predict(test[columns])
sample_submission.to_csv('submission.csv', index=False)
sample_submission.head()
```
