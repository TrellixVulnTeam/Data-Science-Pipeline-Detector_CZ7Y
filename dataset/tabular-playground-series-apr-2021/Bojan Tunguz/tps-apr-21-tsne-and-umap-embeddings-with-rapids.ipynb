```python
import cudf, cuml
import cupy as cp

from cuml.manifold import TSNE, UMAP
import matplotlib.pyplot as plt
from matplotlib.pyplot import ylim, xlim
%matplotlib inline
```

```python
train = pd.read_csv('/kaggle/input/tps-april-2021-feature-eng-only/train.csv')
test = pd.read_csv('/kaggle/input/tps-april-2021-feature-eng-only/test.csv')
```

```python
columns = test.columns
columns
```

```python
train[columns].head()
```

```python
%%time
tsne = TSNE(n_components=2)
train_2D = tsne.fit_transform(train[columns].values)
```

```python
plt.scatter(train_2D[:,0], train_2D[:,1], c=train['Survived'].values, s = 0.8)
```

We see a lot of clustering, but no clear separation between the classes.

```python
%%time
umap = UMAP(n_components=2)
train_2D = umap.fit_transform(train[columns].values)
```

```python
plt.scatter(train_2D[:,0], train_2D[:,1], c=train['Survived'].values, s = 0.8)
```
