```python
!pip install the1owl

import re
import numpy as np
import pandas as pd
from ml import simple

train = pd.read_csv('../input/tabular-playground-series-apr-2021/train.csv')
test = pd.read_csv('../input/tabular-playground-series-apr-2021/test.csv')
```

```python
df = pd.concat((train,test))
df.Age = df.Age.fillna(999)
df.Sex = df.Sex.map({'male':0, 'female':1})
df.Embarked = df.Embarked.map({'S':1, 'C':2, 'Q':3})
df['Cabin2'] = df.Cabin.fillna('').map(lambda x: str(x)[1:] if len(str(x))>1 else 0).astype(int)
df.Cabin = df.Cabin.fillna('').map(lambda x: str(x)[0] if len(str(x))>0 else 9)
df.Cabin = df.Cabin.map({'':9, 'A':1, 'B':2, 'C':3, 'D':4, 'E':5, 'F':6, 'G':7, 'T':8})
#df['Ticket1'] = df.Ticket.map(lambda x: re.sub(r'[0-9./ ]', '', str(x)))
#s = {v:i for i,v in enumerate(sorted(df['Ticket1'].unique()))}
#df['Ticket1'] = df.Ticket1.map(s)
#df['Ticket2'] = df.Ticket.map(lambda x: re.sub(r'[A-Za-z./ ]', '', str(x))).replace('',0).astype(int)

train = df[:len(train)].copy()
test = df[len(train):].copy()
del df
```

```python
data = simple.Data(train, test, 'PassengerId', 'Survived', {'drop':['Name', 'Ticket'], 'fillna':0})
XGB_PARAMS = {'eta': 0.02, 'max_depth': 7, 'objective': 'reg:squarederror', 'eval_metric': ['auc'], 'seed': 21}
model = simple.Model(data, 'XGB', XGB_PARAMS, RANDOM_STATE=21) #MODELS available ['ETR', 'XGB', 'LGB', 'CB']
sub = model.PRED
sub.Survived = (sub.Survived > 0.5).astype(int)
sub.to_csv('submission.csv', index=False)
```
