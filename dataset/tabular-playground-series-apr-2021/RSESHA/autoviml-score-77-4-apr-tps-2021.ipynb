```python
from sklearn.preprocessing import StandardScaler
## First build a simple XGB model to see what baseline performance is like with out any feature engineering from featurewiz
model, features, trainm, testm = Auto_ViML(
    train,
    target,
    test,
    sample_submission=subm,
    hyper_param="RS",
    feature_reduction=True,
    scoring_parameter="weighted-f1",
    KMeans_Featurizer=False,
    Boosting_Flag="CatBoost",
    Binning_Flag=False,
    Add_Poly=1,
    Stacking_Flag=False,
    Imbalanced_Flag=False,
    verbose=1,
)
```

```python
subm[target] = testm[target+'_predictions'].values.astype(int)
subm.head()
```
