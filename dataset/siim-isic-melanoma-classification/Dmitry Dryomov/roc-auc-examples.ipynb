```python
import pandas as pd
import os
DATA_DIR = '../input/siim-isic-melanoma-classification/'

train = pd.read_csv(os.path.join(DATA_DIR, 'train.csv'))
train.shape
```

```python
test = pd.read_csv(os.path.join(DATA_DIR, 'test.csv'))
test.shape
```

```python
total = test.shape[0]
positive = round(total * train.target.mean())
```

```python
from sklearn.metrics import roc_auc_score
```

```python
# 000...00011..11
roc_auc_score([0] * (total - positive) + [1] * (positive), range(total))
```

```python
# 11..11000...000
roc_auc_score([0] * (total - positive) + [1] * (positive), range(total)[::-1])
```

```python
# 00...011..110
roc_auc_score([0] * (total - positive) + [1] * (positive), [total - 1] + list(range(total - 1))) 
```

```python
# 10...0011..11
(roc_auc_score([0] * (total - positive) + [1] * (positive), list(range(total - 1)) + [-1]))
```

```python
# 110...0011..1
(roc_auc_score([0] * (total - positive) + [1] * (positive), list(range(total - 2)) + [-2, -1]))
```

```python
# 1110...0011..1
(roc_auc_score([0] * (total - positive) + [1] * (positive), list(range(total - 3)) + list(range(-3, 0))))
```

```python
# 0...01010101010..101
(roc_auc_score([0] * (total - positive * 2) + [0, 1] * (positive), range(total)))
```

```python
# 0...000001001001001001..001001
(roc_auc_score([0] * (total - positive * 3) + [0, 0, 1] * (positive), range(total)))
```
