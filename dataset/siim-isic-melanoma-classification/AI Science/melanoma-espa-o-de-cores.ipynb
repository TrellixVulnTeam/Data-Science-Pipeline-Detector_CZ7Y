```python
mypath='/kaggle/input/siim-isic-melanoma-classification/jpeg/train/'

from os import listdir
from os.path import isfile, join
image = [mypath+f for f in listdir(mypath) if isfile(join(mypath, f))]
```

```python
image[0]
```

# Lendo imagem com OpenCV

```python
import cv2
image=cv2.imread(image[0])
plt.imshow(image)
```

# Convertendo de BGR para RGB

```python
image_rgb=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image_rgb)
```

# Convers√£o de BGR para HSV

```python
 hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
```

```python
plt.imshow(hsv)
```

# Aplicando Threshold em imagens HSV

```python
MIN = np.array([5, 50, 50],np.uint8)
MAX = np.array([15, 255, 255],np.uint8)


hsv_threshold=cv2.inRange(hsv, MIN, MAX)
```

```python
plt.imshow(hsv_threshold)
```

# Aplicando Threshold a imagem BGR

```python


ret,thresh1 = cv2.threshold(image,127,255,cv2.THRESH_BINARY)
```

```python
plt.imshow(thresh1)
```

```python

ret,thresh1 = cv2.threshold(image,127,255,cv2.THRESH_BINARY_INV)
```

```python
plt.imshow(thresh1)
```
