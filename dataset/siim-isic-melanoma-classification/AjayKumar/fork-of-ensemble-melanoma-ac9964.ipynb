```python
sub = pd.read_csv("../input/siim-isic-melanoma-classification/sample_submission.csv")
del sub['target']

w = [0.05, 0.1, 0.1, 0.1, 0.15, 0.15, 0.15, 0.2]


# w = [0.15, 0.25, 0.2, 0.2, 0.2] ## 9597


##w = [0.85, 0.15]

#w = [0.1, 0.1, 0.1, 0.2, 0.2, 0.2, 0.1]

#w = [0.0, 0.0, 0.0, 0.25, 0.25, 0.25, 0.25] ## 0.955

#w = [0.0, 0.1, 0.1, 0.2, 0.2, 0.2, 0.2]


# w = [0.0, 0.15, 0.2, 0.25, 0.2, 0.2]

```

```python
df.columns = ['image_name', str(counter)]

df[str(counter)] *= w[counter]

df.head()
```

```python
for counter, f in enumerate(FILES):
    df = pd.read_csv(f)
    df.columns = ['image_name', str(counter)]
    df[str(counter)] *= w[counter]
    sub = sub.merge(df, on="image_name")
```

```python
image_name = sub.image_name

sub = sub.drop(columns = ["image_name"])
target = sub.sum(axis = 1)
```
