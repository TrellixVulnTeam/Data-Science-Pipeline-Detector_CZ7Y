```python
import numpy as np
import pandas as pd
import os
```

```python
sub_path = "../input/ensemble-melanoma"
all_files = os.listdir(sub_path)
all_files = [f for f in all_files if 'seresnext50 mean tta 0.9252.csv' not in f and 'b6 2019 mean 0.8666.csv' not in f and 'cpu densenet121 0.8845.csv' not in f]
all_files += ['B3-B6 80 82 size 512.csv', 'triple-stratified-kfold-with-tfrecords 0.9426.csv']
all_files
```

```python
outs = [pd.read_csv(os.path.join(sub_path, f), index_col=0) for f in all_files]
concat_sub = pd.concat(outs, axis=1)
cols = list(map(lambda x: "target" + str(x), range(len(concat_sub.columns))))
concat_sub.columns = cols
concat_sub.reset_index(inplace=True)
concat_sub.head()
ncol = concat_sub.shape[1]
```

```python
concat_sub['target'] = concat_sub.iloc[:, 1:ncol].mean(axis=1)
concat_sub[['image_name', 'target']].to_csv('submission_mean.csv', index=False, float_format='%.6f')
```
