```python
%matplotlib inline
import math
import numpy as np
import matplotlib.pyplot as plt
import os

from skimage import measure
from skimage.color import rgb2gray
from skimage.util import img_as_ubyte
from skimage import io
from skimage.feature import greycomatrix, greycoprops
```

```python
imgList = os.listdir('../input/siimisic-melanoma-512-resized-images/test')
```

```python
imgPath = []
for path in imgList:
    pathImg = f'../input/siimisic-melanoma-512-resized-images/test/{path}'
    imgPath.append(pathImg)
imgPath.sort()
```

```python
ax = []
ay = []
for patch in imgPath:
    image_rgb = io.imread(patch)
    image = img_as_ubyte(rgb2gray(image_rgb))
    glcm = greycomatrix(image, distances=[4], angles=[0], levels=256)
    ax.append(greycoprops(glcm, 'dissimilarity')[0, 0])
    ay.append(greycoprops(glcm, 'correlation')[0, 0])
```

```python
#  4.18879, 2.35619, 3.92699, 1.74533, 3.49066, 5.23599, 5.93412
bx = []
by = []
for patch in imgPath:
    image_rgb = io.imread(patch)
    image = img_as_ubyte(rgb2gray(image_rgb))
    glcm = greycomatrix(image, distances=[4], angles=[0.785398], levels=256)
    bx.append(greycoprops(glcm, 'dissimilarity')[0, 0])
    by.append(greycoprops(glcm, 'correlation')[0, 0])
```

```python
cx = []
cy = []
for patch in imgPath:
    image_rgb = io.imread(patch)
    image = img_as_ubyte(rgb2gray(image_rgb))
    glcm = greycomatrix(image, distances=[4], angles=[1.0472], levels=256)
    cx.append(greycoprops(glcm, 'dissimilarity')[0, 0])
    cy.append(greycoprops(glcm, 'correlation')[0, 0])
```

```python
dx = []
dy = []
for patch in imgPath:
    image_rgb = io.imread(patch)
    image = img_as_ubyte(rgb2gray(image_rgb))
    glcm = greycomatrix(image, distances=[4], angles=[2.0944], levels=256)
    dx.append(greycoprops(glcm, 'dissimilarity')[0, 0])
    dy.append(greycoprops(glcm, 'correlation')[0, 0])
```

```python
ex = []
ey = []
for patch in imgPath:
    image_rgb = io.imread(patch)
    image = img_as_ubyte(rgb2gray(image_rgb))
    glcm = greycomatrix(image, distances=[4], angles=[3.14159], levels=256)
    ex.append(greycoprops(glcm, 'dissimilarity')[0, 0])
    ey.append(greycoprops(glcm, 'correlation')[0, 0])
```

```python
fx = []
fy = []
for patch in imgPath:
    image_rgb = io.imread(patch)
    image = img_as_ubyte(rgb2gray(image_rgb))
    glcm = greycomatrix(image, distances=[4], angles=[4.18879], levels=256)
    fx.append(greycoprops(glcm, 'dissimilarity')[0, 0])
    fy.append(greycoprops(glcm, 'correlation')[0, 0])
```

```python
submission = pd.DataFrame(dict(image_name=imgName, ax=ax, ay=ay, bx=bx, by=by, cx=cx, cy=cy, dx=dx, dy=dy, ex=ex, ey=ey, fx=fx, fy=fy))
submission = submission.sort_values('image_name') 
submission.to_csv('submission.csv', index=False)
submission.head()
```
