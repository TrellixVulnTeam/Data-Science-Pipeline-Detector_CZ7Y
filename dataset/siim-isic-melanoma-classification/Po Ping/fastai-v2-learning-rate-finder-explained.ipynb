```python
import fastai
from fastai.vision.all import *
```

```python
# Defining data path
IMAGE_PATH = "../input/siim-isic-melanoma-classification/"
imgpath = IMAGE_PATH+'/jpeg/train'

df = pd.read_csv('../input/siim-isic-melanoma-classification/train.csv')
#test=pd.read_csv('../input/tweet-sentiment-extraction/test.csv')
df.head()
```

```python
dls = ImageDataLoaders.from_df(df, path=imgpath,valid_pct=0.2, seed=42, fn_col=0, folder=None, suff='.jpg', label_col=7, label_delim=None, y_block=None, valid_col=None, item_tfms=Resize(128), batch_tfms=aug_transforms(flip_vert=True, max_warp=0.), bs=128, val_bs=None, shuffle_train=True)
dls.show_batch()
```

```python
dls.device
```

```python
learn = cnn_learner(dls, resnet34, metrics=error_rate, model_dir="/tmp/model/")
learn.fit_one_cycle(1, 1e-2)
```

```python
learn.lr_find()
```

```python
learn.fine_tune(1)
```
