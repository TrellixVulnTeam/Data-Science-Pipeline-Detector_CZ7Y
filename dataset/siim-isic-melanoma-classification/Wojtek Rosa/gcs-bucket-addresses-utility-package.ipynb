## Initialize kaggle_gcs_client
Before using kaggle_gcs_client please obtain kaggle API key for your account (discussion with organisers [here](https://www.kaggle.com/c/siim-isic-melanoma-classification/discussion/172233)). After this you can set up kaggle API manually, or pass arguments username and key:

```python
from kaggle_secrets import UserSecretsClient
kgc = kaggle_gcs_client(username = 'wrrosa', key = UserSecretsClient().get_secret("kaggleapikey")) 
# optional argument "command", i.e. on my local machine command = "/home/wrosa/.local/bin/kaggle"
```

And simply run:

```python
%%time
kgc.get_gcs_path('cdeotte/melanoma-128x128')
```

By running it for the second time, it will restore cached path:

```python
%%time
kgc.get_gcs_path('cdeotte/melanoma-128x128')
```

Another example:

```python
%%time
kgc.get_gcs_path('cdeotte/melanoma-256x256')
```

## Cache
Now, mechanism behind this magic, is to push notebook to Kaggle and retrieve results. Therefore, for saving time, it's a good practice to first list all of your needed datasets, and then run get_gcs_path():

```python
%%time
kgc = kaggle_gcs_client()
kgc.cache_gcs_paths(['cdeotte/melanoma-128x128','cdeotte/melanoma-256x256','cdeotte/melanoma-384x384','cdeotte/melanoma-512x512'])
```

```python
%%time
print(kgc.get_gcs_path('cdeotte/melanoma-128x128'))
print(kgc.get_gcs_path('cdeotte/melanoma-256x256'))
print(kgc.get_gcs_path('cdeotte/melanoma-384x384'))
print(kgc.get_gcs_path('cdeotte/melanoma-512x512'))
```
