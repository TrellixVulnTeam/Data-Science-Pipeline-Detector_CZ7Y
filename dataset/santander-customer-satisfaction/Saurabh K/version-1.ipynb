```python
traindata_copy = traindata.iloc[:,:-1].copy()

traindata_copy.head()
```

```python
X = np.asarray(traindata.drop(["TARGET","ID"],axis=1))
X = preprocessing.StandardScaler().fit(traindata_copy).transform(traindata_copy)

#Select just Target Column
y = np.asarray(traindata['TARGET'])

```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =.25, stratify = y, random_state = 1)
#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =.20, stratify = y, random_state = 1)
#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =.33, stratify = y, random_state = 1)

```

```python
#clf = DecisionTreeClassifier(criterion = 'gini' , max_depth = 50 , max_leaf_nodes = 5 , min_samples_leaf = 760,splitter = 'best', class_weight = 'balanced')
clf = DecisionTreeClassifier(criterion = 'gini' , max_depth = 40 , max_leaf_nodes = 8, min_samples_leaf = 760, splitter = 'random', class_weight = 'balanced')
#clf = DecisionTreeClassifier(criterion = 'gini' , max_depth = 60 , max_leaf_nodes = 10, min_samples_leaf = 760, splitter = 'random', class_weight = 'balanced')

#clf = DecisionTreeClassifier(criterion = 'entropy' , max_depth = 50 , max_leaf_nodes = 5, min_samples_leaf = 760, splitter = 'random', class_weight = 'balanced')
#clf = DecisionTreeClassifier(criterion = 'entropy' , max_depth = 40 , max_leaf_nodes = 8, min_samples_leaf = 760, splitter = 'random', class_weight = 'balanced')
#clf = DecisionTreeClassifier(criterion = 'entropy' , max_depth = 30 , max_leaf_nodes = 10, min_samples_leaf = 760, splitter = 'random', class_weight = 'balanced')

clf.fit(X_train,y_train)
```

```python
# check accuracy on training set
X_pred_train = clf.predict(X_train)
#Model Accuracy
print("TRAIN Accuracy:", accuracy_score(y_train,X_pred_train))

# check accuracy on test set
X_pred_test = clf.predict(X_test)
#Model Accuracy
print("TEST Accuracy:", accuracy_score(y_test,X_pred_test))




```
