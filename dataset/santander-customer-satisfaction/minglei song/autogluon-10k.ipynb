```python
import pandas as pd
from autogluon.tabular import TabularDataset, TabularPredictor


train_data_76K= TabularDataset('../input/santander-customer-satisfaction/train.csv')
#test_data = TabularDataset('../input/santander-customer-satisfaction/test.csv')

label = 'TARGET'
subsample_size = 10000  # subsample subset of data for faster demo, try setting this to much larger values
train_data_10K = train_data_76K.sample(n=subsample_size, random_state=0)


from sklearn.model_selection import train_test_split


train_data,test_data=train_test_split(train_data_10K)
train_data
```

```python
test_data
```

```python
%%time
metric = 'roc_auc'  # specify your evaluation metric here
predictor = TabularPredictor(label, eval_metric=metric).fit(train_data, presets='best_quality')
```
