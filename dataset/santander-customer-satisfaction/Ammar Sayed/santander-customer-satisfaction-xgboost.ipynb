###  Build Functions

```python
def detect_outliers(df, row, col = 'TARGET'):
    plt.scatter(df[row], df[col])
    plt.xlim((df[row].min(), df[row].max()))


# remove outliers function
def remove_outliers_with_mean(df, col, value):
    outliers = df[bank[col] > value][col].values
    df[col].replace(outliers, df[col].mean(), inplace = True)
    
```

### Split Target from train df

```python
target = bank.iloc[:, -1]
bank.drop('TARGET', axis = 1, inplace = True)
```

### Split Train and Valid also Target

```python
from sklearn.model_selection import train_test_split
x_train, x_valid, y_train, y_valid = train_test_split(bank, target, test_size = 5000)

x_train.shape, x_valid.shape, y_train.shape, y_valid.shape
```

### Scale the features

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

x_train = sc.fit_transform(x_train)
x_valid = sc.transform(x_valid)
x_test = sc.transform(test)
```

### Build the Model

#### XgBoost

```python
import xgboost as xg

xg_clf = xg.XGBClassifier()

xg_clf.fit(x_train, y_train)
```

```python
xg_clf.score(x_valid, y_valid)
```

```python
y_pred_xg = xg_clf.predict(x_test)
y_pred_xg
```

```python
import copy
submit = pd.read_csv(path + 'sample_submission.csv')
y_true = copy.deepcopy(submit['TARGET'])
submit['TARGET'] = y_pred_xg
submit.to_csv('solution_submission.csv', index = False)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_true, y_pred_xg)
```
