con = sqlite3.connect('../input/database.sqlite')

messages = pd.read_sql_query("""
                             SELECT Score, Summary
                             FROM Reviews
                             WHERE Score != 3
                             """, con)

def partition(x):
  if x < 3:
  return 'negative'
return 'positive'

Score = messages['Score']
Score = Score.map(partition)
