```python
cols_list = ["ID", "TARGET"]
x_master = trainData.drop(cols_list, axis=1)
y_master = trainData.loc[:,"TARGET"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_master, y_master, test_size=0.33)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
y_train.head()
```

```python
clf=DecisionTreeClassifier(max_leaf_nodes=10, class_weight="balanced")
```

```python
clf.fit(x_train,y_train)
```

```python
Pred_train = clf.predict(x_train)
```

```python
print("TRAIN Accuracy:", accuracy_score(y_train,Pred_train))
```

```python
Pred_test = clf.predict(x_test)
```

```python
print("TEST Accuracy:", accuracy_score(y_test,Pred_test))
```

```python
roc_auc_score(y_test,Pred_test)
```

```python
roc_auc_score(y_train,Pred_train)
```

```python
print(confusion_matrix(y_test, Pred_test))
print()
```

```python
print('Printing the precision and recall, among other metrics')
print(classification_report(y_train,Pred_train))
```
