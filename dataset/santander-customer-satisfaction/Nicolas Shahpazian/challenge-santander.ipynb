```python
test = pd.read_csv('../input/santander-customer-satisfaction/test.csv')
train = pd.read_csv('../input/santander-customer-satisfaction/train.csv')
```

```python
train
```

Le dataset d'entrainement est composé d'une grande quantité de valeurs numériques et d'une colonne TARGET, qui est la variable à prédire

```python
train['TARGET']
```

```python
train['TARGET'].value_counts()
```

On remarque que la colonne target contient 73012 valeurs '0', et seulement 3008 valeurs '1'. <br>
Cela pose problème car l'algorithme de prédiction risque par la suite d'être biaisé par ce déséquilibre.<br><br>
Il existe plusieurs façons de rééquilibrer un dataset. On choisit ici la méthode smote.

```python
X = train.drop(['TARGET'], axis=1)
y = train['TARGET']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE()
X_train, y_train = smote.fit_resample(X_train, y_train)
y_train.value_counts()
```

# Entrainement du modèle et prédiction #

```python
xgb  = XGBClassifier(use_label_encoder=False)
xgb.fit(X_train, y_train)
```

# Prédiction #

```python
y_xgb = xgb.predict(X_test)
confusion_matrix(y_test, y_xgb)
```

# Rapport #

```python
print(classification_report(y_test, y_xgb))
```

Via XGBClassifier, on obtient alors une précision de 0.92, ce qui est un score relativement satisfaisant.
