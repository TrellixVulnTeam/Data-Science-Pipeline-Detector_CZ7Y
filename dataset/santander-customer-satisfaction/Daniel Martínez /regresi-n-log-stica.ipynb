## División del conjunto de entrenamiento

Se utiliza la estrategia de hold out, la documentación de la librerias utilizadas son: [documentación](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html)



```python
import numpy as np
from sklearn.model_selection import train_test_split
features = ['var3', 'var15', 'imp_ent_var16_ult1', 'imp_op_var39_comer_ult1',
       'imp_op_var39_comer_ult3', 'imp_op_var40_comer_ult1',
       'imp_op_var40_comer_ult3', 'imp_op_var40_efect_ult1',
       'imp_op_var40_efect_ult3','saldo_medio_var33_hace2', 'saldo_medio_var33_hace3',
       'saldo_medio_var33_ult1', 'saldo_medio_var33_ult3',
       'saldo_medio_var44_hace2', 'saldo_medio_var44_hace3',
       'saldo_medio_var44_ult1', 'saldo_medio_var44_ult3', 'var38']
X = train.loc[:,features]
y = train.TARGET
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=0)
```

## Definir el modelo

### Regresión logística
En estadística, la regresión logística es un tipo de análisis de regresión utilizado para predecir el resultado de una variable categórica (una variable que puede adoptar un número limitado de categorías) en función de las variables independientes o predictoras. Es útil para modelar la probabilidad de un evento ocurriendo como función de otros factores. El análisis de regresión logística se enmarca en el conjunto de Modelos Lineales Generalizados (GLM por sus siglas en inglés) que usa como función de enlace la función logit. Las probabilidades que describen el posible resultado de un único ensayo se modelan, como una función de variables explicativas, utilizando una función logística.

Aquí la documentación del modelo: [documentación](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html#sklearn.linear_model.LogisticRegression.score)

```python
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(random_state=0).fit(X_train, y_train)
#clf.predict_proba(X_test)
#error de prueba
print("Error de prueba",clf.score(X_test, y_test))
#error de entrenamiento
print("Error de entrenamiento",clf.score(X_train, y_train))
```

 La metrica aquí presentada es el accuracy promedio.

## Algunas medidas de error

Las medidas a revisar en el siguiente notebook son:
* Accuracy
* Matriz de confusión
* AUC
* F1

### Accuracy

```python
from sklearn.metrics import accuracy_score
print("Error de prueba",accuracy_score(clf.predict(X_test),y_test))
print("Error de prueba",accuracy_score(clf.predict(X_train),y_train))
```

### Matriz de confusión

```python
from sklearn.metrics import confusion_matrix
print("Error de prueba \n",confusion_matrix(clf.predict(X_test),y_test))
print("Error de prueba \n",confusion_matrix(clf.predict(X_train),y_train))
```

### AUC

```python
from sklearn.metrics import roc_auc_score
print("Error de entrenamiento \n",roc_auc_score(y_train, clf.predict_proba(X_train)[:, 1]))
print("Error de prueba \n",roc_auc_score(y_test, clf.predict_proba(X_test)[:, 1]))
```

### F1

```python
from sklearn.metrics import f1_score
print("Error de entrenamiento \n",f1_score(y_train, clf.predict(X_train), average='macro'))
print("Error de prueba \n",f1_score(y_test, clf.predict(X_test), average='macro'))
```
