```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#importing the data
doganella=pd.read_csv("/kaggle/input/acea-water-prediction/Aquifer_Doganella.csv")
auser=pd.read_csv("/kaggle/input/acea-water-prediction/Aquifer_Auser.csv")
amiata=pd.read_csv("/kaggle/input/acea-water-prediction/Water_Spring_Amiata.csv")
bilacino=pd.read_csv("/kaggle/input/acea-water-prediction/Lake_Bilancino.csv")
madonna=pd.read_csv("/kaggle/input/acea-water-prediction/Water_Spring_Madonna_di_Canneto.csv")
luco=pd.read_csv("/kaggle/input/acea-water-prediction/Aquifer_Luco.csv")
petrignano=pd.read_csv("/kaggle/input/acea-water-prediction/Aquifer_Petrignano.csv")
lupa=pd.read_csv("/kaggle/input/acea-water-prediction/Water_Spring_Lupa.csv")
arno=pd.read_csv("/kaggle/input/acea-water-prediction/River_Arno.csv")
```

```python
#function to put the data name to get the results
def eda(data):
    print(data.shape)
    print(data.columns.tolist())
    print(data.describe())
    data[['day','month', 'year']]=data.Date.str.split('/', expand=True)
    print(data.isnull().sum())

    #correlation matrix
    cmap=sns.diverging_palette(220, 10, as_cmap=True)
    plt.figure(figsize=(20,20))
    matrix=np.triu(data.corr())
    sns.heatmap(data.corr(), vmax=1, cmap=cmap, center=0, square=True, linewidths=0.5, cbar_kws={"shrink":0.5}, annot=True, mask=matrix)
    plt.show()

    #average yearly data
    data1=data.drop(['month', 'day', 'Date'], axis=1)
    data1=data1.groupby('year').mean()
    data1=data1.reset_index()

    #average monthly data
    data2=data.drop(['year', 'day', 'Date'], axis=1)
    data2=data2.groupby('month').mean()
    data2=data2.reset_index()

    for i in data2.drop('month', axis=1).columns:
        chart=sns.barplot(data=data2, x=data2.month, y=data2[i])
        plt.show()

    for i in data1.drop('year', axis=1).columns:
        chart=sns.lineplot(data=data1, x=data1.year, y=data1[i])
        chart.set_xticklabels(chart.get_xticklabels(), rotation=90)
        plt.show()
```
