# Aquifer

```python
Aquifer_Auser = pd. read_csv("../input/acea-water-prediction/Aquifer_Auser.csv")
Aquifer_Doganella = pd. read_csv("../input/acea-water-prediction/Aquifer_Doganella.csv")
Aquifer_Luco = pd. read_csv("../input/acea-water-prediction/Aquifer_Luco.csv")
Aquifer_Petrignano = pd. read_csv("../input/acea-water-prediction/Aquifer_Petrignano.csv")
```

<a id="2"></a>
# Aquifer_Rainfall

```python
# Aquifer_Auser-Rainfall
data1 = [col for col in Aquifer_Auser.columns if col.startswith("Rainfall")]
data1.insert(0,"Date")
data1 = Aquifer_Auser[data1]

# Aquifer_Doganella-Rainfall
data2 = [col for col in Aquifer_Doganella.columns if col.startswith("Rainfall")]
data2.insert(0,"Date")
data2 = Aquifer_Doganella[data2]
data2 = data2.rename(columns = {'Date' : 'Date2'})

# Aquifer_Luco-Rainfall
data3 = [col for col in Aquifer_Luco.columns if col.startswith("Rainfall")]
data3.insert(0,"Date")
data3 = Aquifer_Luco[data3]
data3 = data3.rename(columns = {'Date' : 'Date3'})

# Aquifer_Petrignano-Rainfall
data4 = [col for col in Aquifer_Petrignano.columns if col.startswith("Rainfall")]
data4.insert(0,"Date")
data4 = Aquifer_Petrignano[data4]
data4 = data4.rename(columns = {'Date' : 'Date4'})
```

```python
columns1=data1.columns
x_plot = pd.to_datetime(data1['Date'], format = '%d/%m/%Y')
fig, axs = plt.subplots(nrows = len(columns1) - 1, figsize= (22, 6 * len(columns1)))
for var in range(1,len(columns1)):
    axs[var-1].plot(x_plot, data1[columns1[var]])
    axs[var-1].set_xlabel('Date')
    axs[var-1].set_ylabel(columns1[var], fontsize = 17)
    axs[var-1].grid()

columns2=data2.columns
x_plot = pd.to_datetime(data2['Date2'], format = '%d/%m/%Y')
fig, axs = plt.subplots(nrows = len(columns2) - 1, figsize= (25, 6 * len(columns2)))
for var in range(1,len(columns2)):
    axs[var-1].plot(x_plot, data2[columns2[var]])
    axs[var-1].set_xlabel('Date2')
    axs[var-1].set_ylabel(columns2[var], fontsize = 17)
    axs[var-1].grid()

columns3=data3.columns
x_plot = pd.to_datetime(data3['Date3'], format = '%d/%m/%Y')
fig, axs = plt.subplots(nrows = len(columns3) - 1, figsize= (25, 6 * len(columns3)))
for var in range(1,len(columns3)):
    axs[var-1].plot(x_plot, data3[columns3[var]])
    axs[var-1].set_xlabel('Date3')
    axs[var-1].set_ylabel(columns3[var], fontsize = 17)
    axs[var-1].grid()
```
