```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
import os
import cv2
```

Creating Variables for some paths to be used.

```python
input_dir = "../input/cassava-leaf-disease-classification"

train_images_path = os.path.join(input_dir,"train_images")
test_images_path = os.path.join(input_dir,'test_images')
```

Loading Training Data

```python
train = pd.read_csv(os.path.join(input_dir,'train.csv'))
train.head()
```

Plotting a countplot that shows, some the distribution of all classes in the data provided in the data.

```python
sns.countplot(train.label)
plt.show()
```

```python
print("Total Number of Images in Training Data : ",train.shape[0])
```

Checking Class Mappings provided to each class in the train.csv file.

```python
import json

with open(os.path.join(input_dir,"label_num_to_disease_map.json"),'r') as f:
    class_mapping = json.load(f)

class_mapping
```

Below i will be providing some plots showing some samples of images belonging to each class. 
This will just provide an overview of how those images really are and may be hlpeful. 

```python
def plot_samples(class_):
    
    print(f'Some Sample Images belonging to Class {class_mapping[f"{class_}"]}')
    
    sample_images = train[train.label == class_].sample(8)

    fig,ax = plt.subplots(nrows=2,ncols=4,figsize=(16,8))

    for e,img in enumerate(sample_images.image_id):
        image_path = os.path.join(input_dir,f'train_images/{img}')
        image = cv2.imread(image_path)
        ax[e//4][e%4].imshow(image)
    
    plt.show()

plot_samples(0)
```

```python
plot_samples(1)
```

```python
plot_samples(2)
```

```python
plot_samples(3)
```

```python
plot_samples(4)
```

The way we load or read any image matters a lot on our model final results. So here i have a taken a sample image from train_images folder and shown 3 different reading types , namely BGR , RGB and HSV. You can clearly see the difference in these 3 different plots and examine the difference of how can our model can perform on such images.

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import *
image = cv2.imread(os.path.join(input_dir,"train_images/1000015157.jpg"))

fig,ax = plt.subplots(nrows=1,ncols=3,figsize=(21,7))

ax[0].imshow(image)
ax[0].title.set_text("Original Image in BGR Format")

sample_image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
ax[1].imshow(sample_image)
ax[1].title.set_text("Original Image in RGB Format")

sample_image = cv2.cvtColor(sample_image,cv2.COLOR_BGR2HSV)
ax[2].imshow(sample_image)
ax[2].title.set_text("Original Image in HSV Format")
plt.show()
```

```python
fig,ax = plt.subplots(nrows=1,ncols=3,figsize=(21,7))
sns.distplot(image[:,:,0],ax=ax[0],color="b")
sns.distplot(image[:,:,1],ax=ax[1],color="g")
sns.distplot(image[:,:,2],ax=ax[2],color="r")
```

The above plots show distribution of pixel values of a sample image. Channel color is used as repective to the color of the plot.
