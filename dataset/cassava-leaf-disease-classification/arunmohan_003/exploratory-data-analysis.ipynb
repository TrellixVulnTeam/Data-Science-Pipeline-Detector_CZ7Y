In this kernel we will have a quick  Exploratory Data Analysis for Cassava Leaf Disease Classification challenge. 

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from skimage import io
import os
import json
```

```python
base_dir = '/kaggle/input/cassava-leaf-disease-classification/'
train_dir = '/kaggle/input/cassava-leaf-disease-classification/train_images'
test_dir = '/kaggle/input/cassava-leaf-disease-classification/test_images'
```

### Checking size of images

```python
out = []
for img_ in os.listdir(train_dir):
    img = io.imread(os.path.join(train_dir,img_))
    out.append(img.shape)
    

print(set(out))
```

Thus for all train images we have size of 600x800x3

### Traget classes

```python
with open(os.path.join(base_dir, "label_num_to_disease_map.json")) as file:
    map_classes = json.loads(file.read())
    
print(json.dumps(map_classes, indent=4))
```

```python
df = pd.read_csv(os.path.join(base_dir,'train.csv'))
df['target_class'] = df['label'].apply(lambda x: map_classes[str(x)])
df.head()
```

```python
plt.figure(figsize=(18,8))
sns.countplot(df['target_class'])
plt.show()
```

Clearly we have a imabalanced data. So we need to proceed accordingly. Let us viusalize data for each class.

##  Cassava Bacterial Blight (CBB)

```python
# function to plot images
def plot_images(image_name, labels):
    plt.figure(figsize=(18, 14))
    
    for idx, (image_name, label) in enumerate(zip(image_name, labels)):
        plt.subplot(4, 4, idx + 1)
        image = io.imread(os.path.join(train_dir, image_name))
        
        plt.imshow(image)
        plt.title(f"Target Class: {label}", fontsize=10)
        plt.axis("off")
    
```

```python
df_ = df[df["label"] == 0].sample(16)
image_name,labels = df_["image_id"].values,df_["label"].values
plot_images(image_name, labels)
```

## Cassava Brown Streak Disease (CBSD)

```python
df_ = df[df["label"] == 1].sample(16)
image_name,labels = df_["image_id"].values,df_["label"].values
plot_images(image_name, labels)
```

## Cassava Green Mottle (CGM)

```python
df_ = df[df["label"] == 2].sample(16)
image_name,labels = df_["image_id"].values,df_["label"].values
plot_images(image_name, labels)
```

## Cassava Mosaic Disease (CMD)

```python
df_ = df[df["label"] == 3].sample(16)
image_name,labels = df_["image_id"].values,df_["label"].values
plot_images(image_name, labels)
```

## Healthy

```python
df_ = df[df["label"] == 4].sample(16)
image_name,labels = df_["image_id"].values,df_["label"].values
plot_images(image_name, labels)
```
