```python
import numpy as np
import pandas as pd
import lycon
from pathlib import Path
from joblib import Parallel, delayed
from tqdm.notebook import tqdm
```

```python
IMAGE_SIZE = 384
```

```python
def resize_and_save(image_path, target_image_size, target_path):
    """
        1. read image from `image_path`
        2. resize the image to `target_image_size` * `target_image_size`
        3. save the image at `target_path`
    """
    image = lycon.load(image_path)
    image = lycon.resize(image=image, height=target_image_size, width=target_image_size)
    image_name = image_path.split("/")[-1]
    lycon.save(path=f"{target_path}/{image_name}", image=image)
```

```python
%%time
something = Parallel(n_jobs=8, backend="multiprocessing")(
    delayed(resize_and_save)(str(image_path), IMAGE_SIZE, "train_images") for image_path in tqdm(Path("../input/cassava-leaf-disease-classification/train_images/").iterdir())
)
```
