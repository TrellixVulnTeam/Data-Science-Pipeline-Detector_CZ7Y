```python
import numpy as np
import pandas as pd 
from PIL import Image
import matplotlib.pyplot as plt

import json
import os
```

```python
test = '../input/cassava-leaf-disease-classification/test_images'
train = '../input/cassava-leaf-disease-classification/train_images'
train_l = '../input/cassava-leaf-disease-classification/train.csv'
submission = '../input/cassava-leaf-disease-classification/sample_submission.csv'
labels = '../input/cassava-leaf-disease-classification/label_num_to_disease_map.json'
```

```python
data_train = pd.read_csv(train_l)
```

# Вывод доступных классов листьев

```python
with open(labels, 'rb') as f:
    labels_json = json.load(f)

labels_map = {}
for key in labels_json.keys():
    labels_map[int(key)] = labels_json[key]
print(labels_map)
```

```python
data_train.head()
```

# Вывод случайных изображений и классов, к которым они принадлежат

```python
def examples(class_l_num, rows_n=3, cols_n=3):
    class_l = labels_map[class_l_num]
    label_df = data_train[data_train.label == class_l_num]
    rand_idx = np.random.randint(0, len(label_df), rows_n*cols_n)
    fig, axs = plt.subplots(rows_n, cols_n, figsize=(15,15))
    for row in range(rows_n):
        for col in range(cols_n):
            idx = rand_idx[row*cols_n + col]
            img_path = os.path.join(train, label_df.image_id.values[idx])
            img = Image.open(img_path)
            axs[row, col].imshow(img)
            axs[row, col].axis('off')
            axs[row, col].set_title(label_df.image_id.values[idx])
    plt.suptitle(class_l)
    plt.show()
for i in range(len(labels_map)):
    examples(i)
```
