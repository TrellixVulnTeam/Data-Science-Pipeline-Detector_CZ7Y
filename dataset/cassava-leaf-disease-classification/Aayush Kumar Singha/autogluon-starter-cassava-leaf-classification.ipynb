```python
from autogluon import ImageClassification as task
```

```python
dataset = task.Dataset('train_images', label_file='train.csv')
```

```python
test_dataset = task.Dataset('test_images', train=False, scale_ratio_choice=[0.7, 0.8, 0.875])
```

```python
# time_limits = 10 * 60 # 10mins
time_limits = 10 * 60 * 60 * 3
classifier = task.fit(dataset, time_limits=time_limits, verbose=True, epochs=3)
print('Top-1 val acc: %.3f' % classifier.results['best_reward'])
```

```python
inds, probs, probs_all = classifier.predict(test_dataset)
```
