```python
import json
import collections
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
np.random.seed(0)
```

```python
with open('/kaggle/input/cassava-leaf-disease-classification/label_num_to_disease_map.json', 'r') as f:
    json_data = json.load(f)
    print(json.dumps(json_data, indent='\t'))
```

```python
train = pd.read_csv('/kaggle/input/cassava-leaf-disease-classification/train.csv')
print('Number of Train Dataset : {}'.format(len(train)))
```

```python
sort_train = train.label.value_counts().reset_index()
sort_train['disease'] = sort_train['index'].apply(lambda x: str(x)+' : '+json_data[str(x)])

print(collections.Counter(train.label))
sns.barplot(y=sort_train.disease, x=sort_train.label)
plt.xlabel(''); plt.ylabel('')
plt.show()
```

```python
base_dir = '/kaggle/input/cassava-leaf-disease-classification/train_images/'
for i in range(5):
    globals()['label_{}'.format(i)] = train[train.label==i].image_id.tolist()
    plt.figure(figsize=(20, 30))
    plt.suptitle('Label {} : {}'.format(i, json_data['{}'.format(i)]), fontsize=25)
    
    for i, filename in enumerate(np.random.choice(globals()['label_{}'.format(i)], 40)):
        plt.subplot(8, 5, i+1)
        plt.axis('off')
        img = mpl.image.imread(os.path.join(base_dir, filename))
        plt.imshow(img)
    plt.show()
```
