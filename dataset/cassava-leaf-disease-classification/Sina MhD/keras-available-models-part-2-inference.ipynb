## This notebook is part 2 (inference) of the first notebook which was for training <a href='https://www.kaggle.com/sinamhd9/efficientnetb0-in-keras-part-1-training'>[Part 1] </a>

Now we have the trained model, we can simply load it here, predict on the test set and submit the predictions. 

```python
# Importing useful libraries

import pandas as pd 
import numpy as np
import os

import keras
from keras.models import load_model
from PIL import Image
```

## Load model

```python
model = load_model('../input/keras-available-models-part-1-training/model.h5')
```

## Prediction

We can only see 1 data in the test set, however, after submission, all test dataset for public leaderboard (around 15000) will be tested.

```python
img_size = 224
test_images = os.listdir('/kaggle/input/cassava-leaf-disease-classification/test_images/')
y_preds = []

for i in test_images:
    image = Image.open(f'/kaggle/input/cassava-leaf-disease-classification/test_images/{i}')
    image = image.resize((img_size, img_size))
    image = np.expand_dims(image, axis=0)
    y_preds.append(np.argmax(model.predict(image)))
```

```python
df_sub = pd.DataFrame({'image_id': test_images, 'label': y_preds})
display(df_sub)
```

```python
df_sub.to_csv('submission.csv',index=False)
```
