```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Flatten, Dense, Dropout, GlobalAveragePooling2D
from tensorflow.keras.applications.mobilenet import MobileNet, preprocess_input
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image

import numpy as np
import math

import pandas as pd

import os
```

```python
model = tf.keras.models.load_model('../input/cassava-leaf-disease-classification-mobilenet/1605891278')

# Check its architecture
model.summary()
```

```python
test_data = pd.read_csv("../input/cassava-leaf-disease-classification/sample_submission.csv")
```

```python
test_images = test_data['image_id'].values
predictions = []
```

```python
for img_id in test_images:
    img_path = '../input/cassava-leaf-disease-classification/test_images/'+img_id
    img = image.load_img(img_path, target_size=(224, 224))
    img_array = image.img_to_array(img)
    expanded_img_array = np.expand_dims(img_array, axis=0)
    preprocessed_img = expanded_img_array / 255.  # Preprocess the image
    prediction = model.predict(preprocessed_img)
    predicted_class = np.argmax(prediction, axis=-1)[0]
    print(predicted_class)
    predictions.append(predicted_class)
```

```python
preds = np.array(predictions)
preds =preds.reshape(preds.shape[0],-1)
```

```python
test_data['label'] = preds
```

```python
test_data.to_csv("submission.csv",index=False)
```
