```python
import os
import json
import shutil
import pandas as pd
from tqdm.auto import tqdm
from IPython.display import HTML
```

```python
BASE_DIR = "../input/cassava-leaf-disease-classification/"
with open(os.path.join(BASE_DIR, "label_num_to_disease_map.json")) as file:
    classes = json.loads(file.read())
```

```python
classes
```

```python
df = pd.read_csv(os.path.join(BASE_DIR, "train.csv")).set_index('image_id')
```

```python
ls
```

```python
root_name = os.path.join('.', 'images')
if os.path.exists(root_name):
    shutil.rmtree(root_name)
os.makedirs(root_name)
```

```python
# create dir_names and do makedir
dir_names=[]
dir_snames=[]
for k in classes.keys():
    name = os.path.join(root_name, f'{k}')
    dir_names.append(name)
    sname = name + 's'
    dir_snames.append(sname)
    if os.path.exists(name):
        shutil.rmtree(name)
    os.makedirs(name)
    if os.path.exists(sname):
        shutil.rmtree(sname)
    os.makedirs(sname)
```

```python
%%time
# takes about 3 min
for image_name in tqdm(os.listdir(os.path.join(BASE_DIR, "train_images")), desc='[Copy]'):
    label = df.loc[image_name][0] # [0]をつけることでSeriesを値に変換。意味的には最初の一つを取り出している。
    old  = os.path.join(BASE_DIR, 'train_images', image_name)
    new  = os.path.join(dir_names[label], image_name)
    news = os.path.join(dir_snames[label], image_name)
    shutil.copy(old, new)
    shutil.copy(old, news)
```
