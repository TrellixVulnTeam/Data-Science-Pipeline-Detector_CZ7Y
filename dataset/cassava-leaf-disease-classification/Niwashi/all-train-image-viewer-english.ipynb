# 0: Cassava Bacterial Blight (CBB)

[Cassava Bacterial Blight (CBB)](https://en.wikipedia.org/wiki/Bacterial_blight_of_cassava)

In cassava, symptoms vary in a manner that is unique to this pathogen. Symptoms include blight, wilting, dieback, and vascular necrosis. A more diagnostic symptom visible in cassava with X. axonopodis infection are angular necrotic spotting of the leaves—often with a chlorotic ring encircling the spots. These spots begin as distinguishable moist, brown lesions normally restricted to the bottom of the plant until they enlarge and coalesce—often killing the entire leaf. A further diagnostic symptom often embodies itself as pools of gum exudate along wounds and leaf cross veins. It begins as a sappy golden liquid and hardens to form an amber colored deposit.

キャッサバでは、症状はこの病原体に特有の方法で変化します。 症状には、枯死、しおれ、立ち枯れ、血管壊死などがあります。 X. axonopodis感染のキャッサバに見られるより診断的な症状は、葉の角状壊死斑点であり、多くの場合、斑点を取り囲むクロロティックリングがあります。 これらの斑点は、識別可能な湿った茶色の病変として始まり、通常は植物の底に限定され、拡大して合体します。多くの場合、葉全体が死滅します。 さらなる診断症状は、多くの場合、傷や葉の交差静脈に沿って歯茎の滲出液のプールとして現れます。 それは、生意気な金色の液体として始まり、硬化して琥珀色の堆積物を形成します。

```python
n = 0
l_dir = dir_names[n]
s_dir = dir_snames[n]
html_str = (
 '<table border="1">'
  '<tr>')

html_img = ""
html_txt = ""
col_num = 10
for i, img in enumerate(tqdm(sorted(os.listdir(s_dir)), desc=f'[HTML]')):
    l_img = os.path.join(l_dir, img)
    s_img = os.path.join(s_dir, img)
    html_img += f'<td width="64"><a href="{l_img}" title="{img} "target="_blank"><img src="{s_img}"></a></td>'
    base = os.path.splitext(img)[0]
    html_txt += f'<td width="64"><font size="1">{base}</font></td>'
    if not (i+1)%col_num :
        html_str += '</tr>'
        html_str += html_img
        html_img = ""
        html_str += '<tr>'
        html_str += '</tr>'
        html_str += html_txt
        html_txt = ""
        html_str += '<tr>'

html_str += '</tr>'
html_str += html_img
html_img = ""
html_str += '<tr>'
html_str += '</tr>'
html_str += html_txt
html_txt = ""
html_str += '<tr>'
        
html_str += '</tr>'
html_str += '</table>'
```

# 1: Cassava Brown Streak Disease (CBSD)

[Cassava Brown Streak Disease (CBSD)](https://en.wikipedia.org/wiki/Cassava_brown_streak_virus_disease)

CBSD is characterized by severe chlorosis and necrosis on infected leaves, giving them a yellowish, mottled appearance. Chlorosis may be associated with the veins, spanning from the mid vein, secondary and tertiary veins, or rather in blotches unconnected to veins. Leaf symptoms vary greatly depending on a variety of factors. The growing conditions (i.e. altitude, rainfall quantity), plant age, and the virus species account for these differences. Brown streaks may appear on the stems of the cassava plant. Also, a dry brown-black necrotic rot of the cassava tuber exists, which may progress from a small lesion to the whole root. Finally, the roots can become constricted due to the tuber rot, stunting growth. Typically the affected plants do not possess all of these characteristics, but those that are severely affected may. Farmers may be unaware of their infected cassava crops until they are harvested and see the tuber lesions because leaves are asymptomatic. 

CBSDは、感染した葉の重度の白化と壊死を特徴としており、それらに黄色がかった、まだらの外観を与えます。
白化は、中静脈から広がる静脈に関連している可能性があり、二次および三次静脈、またはむしろ静脈に接続されていないしみに関連している可能性があります。葉の症状は、さまざまな要因によって大きく異なります。成長条件（すなわち、高度、降雨量）、植物の年齢、およびウイルス種がこれらの違いを説明します。キャッサバ植物の茎に茶色の縞が現れることがあります。また、キャッサバ塊茎の乾燥した茶黒色の壊死性腐敗が存在し、小さな病変から根全体に進行する可能性があります。最後に、塊茎の腐敗により根が収縮し、発育阻害が起こる可能性があります。通常、影響を受ける植物はこれらの特性のすべてを持っているわけではありませんが、深刻な影響を受ける植物はそうかもしれません。葉は無症候性であるため、農家は、収穫されて塊茎の病変が見えるまで、感染したキャッサバ作物に気付かない可能性があります

```python
n = 1
l_dir = dir_names[n]
s_dir = dir_snames[n]
html_str = (
 '<table border="1">'
  '<tr>')

html_img = ""
html_txt = ""
col_num = 10
for i, img in enumerate(tqdm(sorted(os.listdir(s_dir)), desc=f'[HTML]')):
    l_img = os.path.join(l_dir, img)
    s_img = os.path.join(s_dir, img)
    html_img += f'<td width="64"><a href="{l_img}" title="{img} "target="_blank"><img src="{s_img}"></a></td>'
    base = os.path.splitext(img)[0]
    html_txt += f'<td width="64"><font size="1">{base}</font></td>'
    if not (i+1)%col_num :
        html_str += '</tr>'
        html_str += html_img
        html_img = ""
        html_str += '<tr>'
        html_str += '</tr>'
        html_str += html_txt
        html_txt = ""
        html_str += '<tr>'

html_str += '</tr>'
html_str += html_img
html_img = ""
html_str += '<tr>'
html_str += '</tr>'
html_str += html_txt
html_txt = ""
html_str += '<tr>'
        
html_str += '</tr>'
html_str += '</table>'
```

# 2: Cassava Green Mottle (CGM)

[Cassava Green Mottle (CGM)](http://www.pestnet.org/fact_sheets/cassava_green_mottle_068.htm)

Young leaves are puckered with faint to distinct yellow spots, green patterns (mosaics), and twisted margins. Usually, the shoots recover from symptoms and appear healthy. Occasionally, plants become severely stunted, edible roots are absent or, if present, they are small and woody when cooked. Detection & Inspection: Look for yellow patterns on the leaves, from small dots to irregular patches of yellow and green. Look for leaf margins that are distorted. The plants may be stunted.

若い葉は、かすかにはっきりとした黄色の斑点、緑色の模様（モザイク）、ねじれた縁でしわが寄っています。通常、新芽は症状から回復し、健康に見えます。時折、植物はひどく発育不全になり、食用の根がないか、存在する場合は、調理したときに小さくて木質になります。検出と検査:小さな点から黄色と緑の不規則なパッチまで、葉の黄色のパターンを探します。歪んでいる葉の余白を探します。植物は発育不全になる可能性があります。
