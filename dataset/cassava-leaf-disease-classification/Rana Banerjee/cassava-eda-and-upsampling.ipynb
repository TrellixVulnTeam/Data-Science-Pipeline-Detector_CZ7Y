# Upsampling the data to balance the dataset
### Like Jeremy has mentioned in fastai and otherwise also agreed that the simplest way of just duplicating the minority classes as an upsampling strategy works pretty well. On the same lines I am sharing my eda and the upsampled code data as below.

```python
train_df = pd.read_csv('/kaggle/input/cassava-leaf-disease-classification/train.csv')
```

```python
tdf_counts = list(train_df.groupby('label').count()['image_id'])
```

```python
max_count = max(tdf_counts)
max_count
```

```python
tdf_copy = [(int(max_count/x), max_count-x*int(max_count/x)) for x in tdf_counts]
```

```python
for i in range(len(tdf_counts)):
    mul, rem = tdf_copy[i]
    new_df = train_df[train_df['label']==i]
    for j in range(mul-1):
        train_df=pd.concat([train_df,new_df]).reset_index(drop=True)
    new_df = new_df.sample(n=rem)
    train_df=pd.concat([train_df,new_df]).reset_index(drop=True)
```
