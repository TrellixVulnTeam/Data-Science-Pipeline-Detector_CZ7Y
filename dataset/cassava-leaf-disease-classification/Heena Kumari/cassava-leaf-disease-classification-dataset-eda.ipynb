# Importing necessary libraries

```python
import numpy as np
import pandas as pd
import json
import matplotlib.pyplot as plt
from PIL import Image
import os
```

# Training Data Analysis

### Loading Training Images ID and labels file

```python
df=pd.read_csv('/kaggle/input/cassava-leaf-disease-classification/train.csv')

# Adding Class Label Name Column
with open('../input/cassava-leaf-disease-classification/label_num_to_disease_map.json') as file:
    cls_names_dict=json.load(file)
cls_names=[]
cls_names=[cls_names_dict[str(i)] for i in df['label']]
df['cls_names']=cls_names
df.head(20)
```

## Class-wise Images count

```python
df.groupby('cls_names').count()
```

# Displaying different class Images

```python
healthy_images=df[df['label']==0]
plt.figure(figsize=(16,20))
for i in range(50):
    plt.subplot(10,5,i+1)
    plt.imshow(Image.open('/kaggle/input/cassava-leaf-disease-classification/train_images/'+healthy_images['image_id'].iloc[i]))
    #plt.ylabel(healthy_images['image_id'].iloc[i])
    plt.axis("off")
```
