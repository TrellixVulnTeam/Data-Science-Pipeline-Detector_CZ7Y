# **Do some data augmentation experiment with Albumentations API**

```python
import numpy as np
import os
import cv2
from matplotlib import pyplot as plt
```

```python
from albumentations import (
    Blur, CLAHE, ChannelDropout, ChannelShuffle, ColorJitter, Downscale,
    Equalize, FDA, FancyPCA, GaussNoise, GaussianBlur, GlassBlur, HistogramMatching,
    HueSaturationValue, IAAAdditiveGaussianNoise, IAAEmboss, IAASharpen, IAASuperpixels,
    ISONoise, MedianBlur, MotionBlur, MultiplicativeNoise, Normalize, RGBShift,
    RandomBrightnessContrast, RandomFog, RandomGamma, RandomRain, RandomShadow, RandomSnow, RandomSunFlare,
    CenterCrop, CoarseDropout, Crop, CropNonEmptyMaskIfExists, ElasticTransform, Flip, GridDistortion, GridDropout,
    HorizontalFlip, IAAAffine, IAACropAndPad, IAAFliplr, IAAFlipud, IAAPerspective, IAAPiecewiseAffine, Lambda,
    LongestMaxSize, MaskDropout, NoOp, OpticalDistortion, PadIfNeeded, RandomCrop, RandomCropNearBBox, RandomGridShuffle,
    RandomResizedCrop, RandomRotate90, RandomScale, RandomSizedBBoxSafeCrop, RandomSizedCrop, Resize, Rotate, ShiftScaleRotate,
    SmallestMaxSize, Transpose, VerticalFlip
)  # 图像变换函数
```

```python
image = cv2.imread('../input/cassava-leaf-disease-classification/train_images/1000723321.jpg', 1)  # BGR
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
def testAlbumentations(image):
    '''Test Albumentations API'''
    '''Spatial-level transforms'''

    aug = Blur(p=1)
    img_Blur = aug(image=image)['image']# 可以

    aug = CLAHE(p=1)
    img_CLAHE = aug(image=image)['image']# 需要

    aug = ChannelDropout(p=1)
    img_ChannelDropout = aug(image=image)['image']# 不用吧

    aug = ChannelShuffle(p=1)
    img_ChannelShuffle = aug(image=image)['image']# 不用吧

    aug = ColorJitter(p=1)
    img_ColorJitter = aug(image=image)['image']# 可以考虑

    aug = Downscale(p=1)
    img_Downscale = aug(image=image)['image']# 不用吧

    aug = Equalize(p=1)
    img_Equalize = aug(image=image)['image']# 不用吧

    # domain_adaptation.FDA 可以研究

    aug = FancyPCA(p=1)
    img_FancyPCA = aug(image=image)['image']# 可以

    # FromFloat

    aug = GaussNoise(p=1)
    img_GaussNoise = aug(image=image)['image']# 可以

    aug = GaussianBlur(p=1)
    img_GaussianBlur = aug(image=image)['image']# 可以

    aug = GlassBlur(p=1)
    img_GlassBlur = aug(image=image)['image']# 可以

    # aug = HistogramMatching(p=1)
    # img_HistogramMatching = aug(image=image)['image']

    aug = HueSaturationValue(p=1)
    img_HueSaturationValue = aug(image=image)['image']# 可以

    aug = IAAAdditiveGaussianNoise(p=1)
    img_IAAAdditiveGaussianNoise = aug(image=image)['image']# 可以

    aug = IAAEmboss(p=1)
    img_IAAEmboss = aug(image=image)['image']# 可以

    aug = IAASharpen(p=1)
    img_IAASharpen = aug(image=image)['image']# 可以

    aug = IAASuperpixels(p=1)
    img_IAASuperpixels = aug(image=image)['image']# ？

    aug = ISONoise(p=1)
    img_ISONoise = aug(image=image)['image']# 可以

    aug = MedianBlur(p=1)
    img_MedianBlur = aug(image=image)['image']# 可以 (用像素点邻域灰度值的中值来代替该像素点的灰度值，让周围的像素值接近真实的值从而消除孤立的噪声点

    aug = MotionBlur(p=1)
    img_MotionBlur = aug(image=image)['image']# 不用吧

    aug = MultiplicativeNoise(p=1)
    img_MultiplicativeNoise = aug(image=image)['image']# 不用吧

    aug = Normalize(mean=(0.43032044, 0.49672633, 0.31342009), std=(0.21909042, 0.22394303, 0.20059192), p=1)
    img_Normalize = aug(image=image)['image']# 需要

    aug = RGBShift(p=1)
    img_RGBShift = aug(image=image)['image']# 可以

    aug = RandomBrightnessContrast(p=1)
    img_RandomBrightnessContrast = aug(image=image)['image']# 需要

    aug = RandomFog(p=1)
    img_RandomFog = aug(image=image)['image']# 可以

    aug = RandomGamma(p=1)
    img_RandomGamma = aug(image=image)['image']# 可以

    aug = RandomRain(p=1)
    img_RandomRain = aug(image=image)['image']# 可以

    aug = RandomShadow(p=1)
    img_RandomShadow = aug(image=image)['image']# 可以

    aug = RandomSnow(p=1)
    img_RandomSnow = aug(image=image)['image']# 不用

    aug = RandomSunFlare(p=1)
    img_RandomSunFlare = aug(image=image)['image']# 不用

    '''Spatial-level transforms'''
    # CenterCrop
    aug = ElasticTransform(p=1)
    img_ElasticTransform = aug(image=image)['image']# 可以

    aug = CoarseDropout(fill_value=(0.43032044, 0.49672633, 0.31342009), p=1)
    img_CoarseDropout = aug(image=image)['image']# 需要

    aug = GridDistortion(p=1)
    img_GridDistortion = aug(image=image)['image']# 需要

    aug = GridDropout(p=1)
    img_GridDropout = aug(image=image)['image']# 不用

    aug = IAAAffine(p=1)
    img_IAAAffine = aug(image=image)['image']# 可以

    aug = IAAPerspective(p=1)
    img_IAAPerspective = aug(image=image)['image']# 可以

    aug = IAAPiecewiseAffine(p=1)
    img_IAAPiecewiseAffine = aug(image=image)['image']# 可以

    aug = OpticalDistortion(p=1)
    img_OpticalDistortion = aug(image=image)['image']# 可以

    aug = RandomGridShuffle(p=1)
    img_RandomGridShuffle = aug(image=image)['image']# 可以

    '''空间几何变换类'''

    # 仿射变换

    # 视觉变换

    # 分段仿射

    '''像素颜色变换类'''

    '''多样本合成类'''

    # show
    plt.subplots_adjust(hspace=0.4)
    plt.axis('off')
    plt.subplot(2, 2, 1)
    plt.imshow(image)
    plt.title("original")
    plt.subplot(2, 2, 2)
    plt.imshow(img_Blur)
    plt.title("Blur")
    plt.subplot(2, 2, 3)
    plt.imshow(img_CLAHE)
    plt.title("CLAHE")
    plt.subplot(2, 2, 4)
    plt.imshow(img_ColorJitter)
    plt.title("ColorJitter")
    plt.show()
    
    plt.subplots_adjust(hspace=0.4)
    plt.subplot(2, 2, 1)
    plt.imshow(img_GaussianBlur)
    plt.title("GaussianBlur")
    plt.subplot(2, 2, 2)
    plt.imshow(img_Equalize)
    plt.title("Equalize")
    plt.subplot(2, 2, 3)
    plt.imshow(img_FancyPCA)
    plt.title("FancyPCA")
    plt.subplot(2, 2, 4)
    plt.imshow(img_GaussNoise)
    plt.title("GaussNoise")
    plt.show()

    plt.subplots_adjust(hspace=0.4)
    plt.subplot(2, 2, 1)
    plt.imshow(img_GlassBlur)
    plt.title("GlassBlur")
    plt.subplot(2, 2, 2)
    plt.imshow(img_ISONoise)
    plt.title("ISONoise")
    plt.subplot(2, 2, 3)
    plt.imshow(img_HueSaturationValue)
    plt.title("HueSaturationValue")
    plt.subplot(2, 2, 4)
    plt.imshow(img_IAAAdditiveGaussianNoise)
    plt.title("IAAAdditiveGaussianNoise")
    plt.show()

    plt.subplots_adjust(hspace=0.4)
    plt.subplot(2, 2, 1)
    plt.imshow(img_IAAEmboss)
    plt.title("IAAEmboss")
    plt.subplot(2, 2, 2)
    plt.imshow(img_IAASharpen)
    plt.title("IAASharpen")
    plt.subplot(2, 2, 3)
    plt.imshow(img_IAASuperpixels)
    plt.title("IAASuperpixels")
    plt.subplot(2, 2, 4)
    plt.imshow(img_MedianBlur)
    plt.title("MedianBlur")
    plt.show()

    plt.subplots_adjust(hspace=0.4)
    plt.subplot(2, 2, 1)
    plt.imshow(img_MotionBlur)
    plt.title("MotionBlur")
    plt.subplot(2, 2, 2)
    plt.imshow(img_MultiplicativeNoise)
    plt.title("MultiplicativeNoise")
    plt.subplot(2, 2, 3)
    plt.imshow(img_RandomBrightnessContrast)
    plt.title("RandomBrightnessContrast")
    plt.subplot(2, 2, 4)
    plt.imshow(img_RGBShift)
    plt.title("RGBShift")
    plt.show()

    plt.subplots_adjust(hspace=0.4)
    plt.subplot(2, 2, 1)
    plt.imshow(img_RandomFog)
    plt.title("RandomFog")
    plt.subplot(2, 2, 2)
    plt.imshow(img_RandomGamma)
    plt.title("RandomGamma")
    plt.subplot(2, 2, 3)
    plt.imshow(img_RandomRain)
    plt.title("RandomRain")
    plt.subplot(2, 2, 4)
    plt.imshow(img_RandomShadow)
    plt.title("RandomShadow")
    plt.show()

    plt.subplots_adjust(hspace=0.4)
    plt.subplot(2, 2, 1)
    plt.imshow(img_ElasticTransform)
    plt.title("ElasticTransform")
    plt.subplot(2, 2, 2)
    plt.imshow(img_CoarseDropout)
    plt.title("CoarseDropout")
    plt.subplot(2, 2, 3)
    plt.imshow(img_GridDistortion)
    plt.title("GridDistortion")
    plt.subplot(2, 2, 4)
    plt.imshow(img_IAAAffine)
    plt.title("IAAAffine")
    plt.show()

    plt.subplots_adjust(hspace=0.4)
    plt.subplot(2, 2, 1)
    plt.imshow(img_IAAPerspective)
    plt.title("IAAPerspective")
    plt.subplot(2, 2, 2)
    plt.imshow(img_IAAPiecewiseAffine)
    plt.title("IAAPiecewiseAffine")
    plt.subplot(2, 2, 3)
    plt.imshow(img_OpticalDistortion)
    plt.title("OpticalDistortion")
    plt.subplot(2, 2, 4)
    plt.imshow(img_RandomGridShuffle)
    plt.title("RandomGridShuffle")
    plt.show()
```
