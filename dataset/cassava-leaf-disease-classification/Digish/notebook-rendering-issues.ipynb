```python
import tensorflow as tf
import pandas as pd
import matplotlib.image as image
import os
import itertools
```

```python
'''Utility Methods and objects needed for writing data to file'''
def _bytes_feature(value):
  """Returns a bytes_list from a string / byte."""
  if isinstance(value, type(tf.constant(0))):
    value = value.numpy() # BytesList won't unpack a string from an EagerTensor.
  return tf.train.Feature(bytes_list=tf.train.BytesList(value=[value]))

def _float_feature(value):
  """Returns a float_list from a float / double."""
  return tf.train.Feature(float_list=tf.train.FloatList(value=[value]))

def _int64_feature(value):
  """Returns an int64_list from a bool / enum / int / uint."""
  return tf.train.Feature(int64_list=tf.train.Int64List(value=[value]))

def convert_to_image_example(image_string,image_name, target):
    feature = {
      'image': _bytes_feature(image_string),
      'image_name': _bytes_feature(image_name),        
      'target': _int64_feature(target)
    }
    return tf.train.Example(features=tf.train.Features(feature=feature))


def write_file(dataset,outputfilename):
    record_file = outputfilename
    base_image_path = '../input/cassava-leaf-disease-classification/train_images/'
    with tf.io.TFRecordWriter(record_file) as writer:
        for row in dataset.itertuples(index=False):
            img = open(base_image_path + row[0], 'rb').read()
            tf_example = convert_to_image_example(img,row[0].encode(),row[1])
            writer.write(tf_example.SerializeToString())
            
            
```

```python
overall_dataset = pd.read_csv('../input/cassava-leaf-disease-classification/train.csv')
sampleds = overall_dataset[0:1]
write_file(sampleds,'-1_generated.tfrec')
#write_file(sampleds,'1_generated.tfrec')
```
