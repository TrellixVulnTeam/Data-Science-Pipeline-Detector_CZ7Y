# Analyze your model performance by confusion matrix

```python
import numpy as np
import pandas as pd
```

```python
# inference results
df_train = pd.read_csv("../input/inference5folds/train_inference_5folds.csv")

df_train.head()
```

```python
# read logits
logits = np.array([df_train["logits0"], df_train["logits1"], df_train["logits2"], df_train["logits3"] ,df_train["logits4"]]).T
logits
```

```python
# accuracy for 5-fold
np.sum((df_train.label.values==logits.argmax(1)))/(len(df_train))*100
```

```python
# get df for incorrect predictions
df_train[df_train["label"]!=logits.argmax(1)]
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(df_train["label"], logits.argmax(1))

print(cm)
```

```python
for i, val in enumerate(cm):
    print("for class {}: accuracy: {}".format(i, val[i]/sum(val)*100))
```

```python
for i, val in enumerate(cm[:-1]):
    print("for class {}: possibility to mistake for healthy: {}".format(i, val[4]/val[i]*100))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(cm)
plt.savefig('.sklearn_confusion_matrix.png')
```

# Calculate P, R, f1

```python
from sklearn.metrics import classification_report

print(classification_report(df_train["label"], logits.argmax(1)))
```
