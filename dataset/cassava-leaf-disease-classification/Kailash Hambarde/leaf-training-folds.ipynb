```python
import pandas as pd
from sklearn.model_selection import StratifiedKFold
import os
```

# Read Data

```python
df = pd.read_csv("../input/cassava-leaf-disease-classification/train.csv")
```

Create k fold using Stratified KFold

```python
df["kfold"] = -1
#randomly shafl data
df = df.sample(frac=1).reset_index(drop=True)
y = df.label.values
skf = StratifiedKFold(n_splits=5)
```

```python
df.head()
```

```python
for f, (t_, v_) in enumerate(skf.split(X=df, y = y)):
    #filed the validation with current fold number
    df.loc[v_, "kfold"] = f
```

```python
df.head()
```

```python
image_path = "../cassave-leaf_disease-classification/train_images"
df["image_path"] = [os.path.join(image_path, x)for x in df.image_id.values]
df['image_path']
df.to_csv('./trainfold.csv')
```
