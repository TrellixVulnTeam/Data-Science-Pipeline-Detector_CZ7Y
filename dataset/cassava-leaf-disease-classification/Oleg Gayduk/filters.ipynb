```python
f = 3
p = 0
s = 1

n = len(im)

final_size = int((n + 2 * p - f)/s + 1) # in this case it's equal to n - f + 1 (cause of p = 0 and s = 1)
```

```python
n, m = final_size, final_size
arr = [[[0 for p in range(f)] for j in range(m)] for i in range(n)]

u, g, summ = 0, 0, 0

def func(i):
    for k in range(f):
            for l in range(f):
                summ += im[l + u][k + g][0] * f1[l][k]
    return summ

for i in range(final_size):
    for j in range(final_size):
        #R range
        for k in range(f):
            for l in range(f):
                summ += im[l + u][k + g][0] * f1[l][k]

        arr[i][j][0] = summ 
        
        summ = 0

        #G range
        for k in range(f):
            for l in range(f):
                summ += im[l + u][k + g][1] * f1[l][k]
                
        arr[i][j][1] = summ 
        
        summ = 0
        
        #B range
        for k in range(f):
            for l in range(f):
                summ += im[l + u][k + g][2] * f1[l][k]
        
        arr[i][j][2] = summ
        
        g += s
        if(g >= final_size):
            u += s
            g = 0
            if(u >= final_size):
                u = 0
        
#print(arr)

cv2.imwrite("test.jpg", np.array(arr))
```
