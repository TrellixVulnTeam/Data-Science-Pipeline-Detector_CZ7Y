# Cassava Leaf Disease Classification

Challenge is [HERE](https://www.kaggle.com/c/cassava-leaf-disease-classification)

## Data

* Train Image Resolution : (500 ~ 800)x(500 ~ 800)
* Test Image Resolution : 800x600
* Total 5 classes (4 for the diseases, 1 for healthy)
* There are lots of noisy labels (both train & test)

### Label Distribution

* cleaned 19 + 20 datasets : [here](https://www.kaggle.com/kozistr/leaf-disease-cleaned)

% cleaned label : pseudo label (got from my best lb models, 0.905)

## Works

1. extra data (using 2019 + 2020 data)
2. TTA (n_iter = 4 is best)
3. smooth cross entropy loss (maybe...?)
4. heavy augmentations (brightness, contrastive, flip, etc...) (?)
5. `CutMix` + `FMix` (?)

## Not Works

1. `SnapMix` Augmentations
2. too small or big backbone (ResNeXt50~, EffNet-B5~ )
3. bi-tempered loss
4. taylor category cross entropy loss
5. some augmentations (e.g. ChannelDropout, GridDistortion, RGDShift)
6. tuning on 2020 validation dataset
