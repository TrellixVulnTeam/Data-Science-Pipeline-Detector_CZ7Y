```python
import csv
import os
import numpy as np
import shutil
```

```python
a_train_label = '../input/labels/train_label.csv'
a_test_label = '../input/labels/test_label.csv'
```

```python
'''将图片按照label文件分别拷贝进train_image和test_image目录内'''
all_image = '../input/cassava-leaf-disease-classification/train_images'
train_image = './train_image'
test_image = './test_image'
os.mkdir(train_image)
os.mkdir(test_image)
tr_label = []
te_label = []
train_data = []
test_data = []
```

```python
# 读取csv数据
with open(a_train_label)as train_file:
    train_reader = csv.reader(train_file)
    tr_label = next(train_reader)
    for row in train_reader:
        train_data.append(row)
with open(a_test_label)as test_file:
    test_reader = csv.reader(test_file)
    te_label = next(test_reader)
    for row in test_reader:
        test_data.append(row)

# 生成训练数据集
for row in train_data:
    shutil.copyfile(os.path.join(all_image, row[0]),os.path.join(train_image, row[0]))


# 生成测试数据集
for row in test_data:
    shutil.copyfile(os.path.join(all_image, row[0]),os.path.join(test_image, row[0]))
```

```python
shutil.copyfile('../input/labels/test_label.csv' , './test_labels.csv')
shutil.copyfile('../input/labels/train_label.csv' , './train_labels.csv')
```
