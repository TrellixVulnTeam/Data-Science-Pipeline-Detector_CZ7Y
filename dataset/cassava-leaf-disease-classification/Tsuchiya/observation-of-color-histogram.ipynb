```python
import pandas as pd
import numpy as np
import json

from PIL import Image
import matplotlib.pyplot as plt
```

```python
base_path = '/kaggle/input/cassava-leaf-disease-classification/'
train_img_path = base_path + 'train_images/'
test_img = '/kaggle/input/cassava-leaf-disease-classification/test_images/2216849948.jpg'
```

```python
sample_submission = pd.read_csv(base_path + 'sample_submission.csv')

label_num_to_disease_map = open(base_path + 'label_num_to_disease_map.json', 'r')
label_num_to_disease_map = json.load(label_num_to_disease_map)

train = pd.read_csv(base_path + 'train.csv')
```

```python
def show_color_hist(label, pick_count=20):
    img_filename_list = train.loc[train['label']==label, 'image_id'].to_list()
    
    for f in img_filename_list[:pick_count]:
        
        img_path = train_img_path + f

        fig, (axL, axR) = plt.subplots(ncols=2, figsize=(10,4))

        im = Image.open(img_path)

        im_array = np.asarray(im)
        axL.imshow(im_array)

        im_array_rgb = np.asarray(im.convert("RGB")).reshape(-1,3)
        axR.hist(im_array_rgb, color=["red", "green", "blue"], histtype="step", bins=256)

        fig.show()
```

# Cassava Bacterial Blight (CBB)

```python
show_color_hist(0)
```

# Cassava Brown Streak Disease (CBSD)

```python
show_color_hist(1)
```

# Cassava Green Mottle (CGM)

```python
show_color_hist(2)
```

# Cassava Mosaic Disease (CMD)

```python
show_color_hist(3)
```
