# Cassava Leaf Disease - Data Analysis
In Cassava Leaf disease competition, They provided dataset of 21,367 labeled images. Most images were crowdsourced from farmers taking photos of their gardens.

Using this dataset we are going to explore images and see insights of data to get better understanding of problem.

In this competition, We need to solve multi class classification problem where we need to classify plant image to any of 5 classes(4 diseases or healthy plant).

```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import json
from math import sqrt
import os
```

```python
BASE_DIR = '/kaggle/input/cassava-leaf-disease-classification/'
train_image_base_dir = BASE_DIR + 'train_images'
filenames = os.listdir(train_image_base_dir)
print(filenames[0])
```

```python
# we will use it later
def plot_images(df, no_of_images):
    """
    function that will plot images for visualisation
    """
    
    size = int(sqrt(no_of_images))
    fig, axis = plt.subplots(size, size)
    fig.set_figheight(15)
    fig.set_figwidth(15)
    for i in range(size * size):
        image = plt.imread(os.path.join(train_image_base_dir, df.loc[i, 'image_id']))
        axis[int(i/size)][int(i%size)].imshow(image)
        axis[int(i/size)][int(i%size)].title.set_text(df.loc[i, 'image_id'])
    plt.show()
```

# General Visualization of Images

```python
train_label_file_path = BASE_DIR + 'train.csv'
train_df = pd.read_csv(train_label_file_path)
print(train_df.loc[0:8])
print('Total no of training Images: ', len(train_df))
plot_images(train_df, no_of_images=9)
```

## Multiclass classification based on 5 classes (4 diseases and healthy plant)

```python
with open(BASE_DIR + 'label_num_to_disease_map.json') as f:
    label_mapping = json.load(f)
print(json.dumps(label_mapping, indent=4, sort_keys=True))
```

## Images of leaves having Cassava Bacterial Blight(CBB) with label 0

```python
ccb_df = train_df[train_df['label'] == 0].reset_index()
print('No of Images of Cassava Bacterial Blight (CBB) in training set:', len(ccb_df))
plot_images(ccb_df, no_of_images=4)
```

## Images of leaves having Cassava Brown Streak Disease (CBSD) with label 1

```python
cbsd_df = train_df[train_df['label'] == 1].reset_index()
print('No of images of Cassava Brown Streak Disease (CBSD) for training:', len(cbsd_df))
plot_images(cbsd_df, no_of_images=4)
```

## Images of leaves having Cassava Green Mottle (CGM) with label 2

```python
cgm_df = train_df[train_df['label'] == 2].reset_index()
print('No of Images for Cassava Brown Streak Disease (CBSD) in training set:', len(cgm_df))
plot_images(cgm_df, no_of_images=4)
```

## Images of leaves having Cassava Mosaic Disease (CMD) with label 3

```python
cmd_df = train_df[train_df['label'] == 3].reset_index()
print('No of Images in Cassava Mosaic Disease (CMD) for training:', len(cmd_df))
plot_images(cmd_df, no_of_images=4)
```

## Images of healthy plants

```python
hp_df = train_df[train_df['label'] == 4].reset_index()
print('No of Images of Helthy plants:', len(hp_df))
plot_images(hp_df, no_of_images=4)
```
