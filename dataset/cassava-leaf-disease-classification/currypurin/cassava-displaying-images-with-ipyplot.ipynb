```python
import ipyplot
import pandas as pd
import numpy as np
from PIL import Image, ImageFilter
```

```python
ls ../input/
```

```python
train = pd.read_csv('../input/cassava-leaf-disease-classification/train.csv')
train.head()
```

```python
ls ../input/cassava-leaf-disease-classification/
```

```python
import os

train_image_dir = '../input/cassava-leaf-disease-classification/train_images/'
SAMPLE_NUM = 20
images_path = []
images_pil = []
images_ndarray = []


for i in range(SAMPLE_NUM):
    images_path.append(train_image_dir + train['image_id'].values[i])
    image_pil = Image.open(images_path[i])
    images_pil.append(image_pil)
    images_ndarray.append(np.array(image_pil))
```

# PLOT CRASS TABS 

```python
ipyplot.plot_class_tabs(images_path, train['label'].values[:SAMPLE_NUM], force_b64=True)
# The kaggle notepad and jupyter notebook environment, seem to require force_b64 to be specified.  ref:https://github.com/karolzak/ipyplot/issues/22
```

```python
ipyplot.plot_class_tabs(images_pil, train['label'].values[:SAMPLE_NUM])
```

```python
ipyplot.plot_class_tabs(images_ndarray, train['label'].values[:SAMPLE_NUM])
```

# plot images

```python
ipyplot.plot_images(images_pil, train['label'].values[:SAMPLE_NUM])
```

# plot_class_representations

```python
ipyplot.plot_class_representations(images_pil, train['label'].values[:SAMPLE_NUM])
```
