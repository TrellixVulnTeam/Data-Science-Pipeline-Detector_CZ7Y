# 2.Disease details

2.1 CCB:
Main characteristics to leverage: angular spots, brown spots with yellow borders, yellow leaves, leaves wilting


2.2 CBSD
Main characteristics to leverage:The infected leaves do not become distorted in shape as occurs with leaves infected by Cassava mosaic disease.

2.3 CGM
Main characteristics to leverage:yellow patterns, irregular patches of yellow and green, leaf margins distortion, stunted

2.4 CMD
Main characteristics to leverage: severe shape distortion, mosaic patterns

There is a strong coupling between the characteristics of the disease, and it is necessary to find more effective classification features
