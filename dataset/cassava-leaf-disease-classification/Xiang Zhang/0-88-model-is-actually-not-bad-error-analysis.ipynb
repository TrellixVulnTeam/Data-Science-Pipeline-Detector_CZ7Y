As I desperate to make a little tiny improvement in public leaderboard, it comes to my mind that I should check how is my poor model actually doing. My model's accuracy is about 0.87-0.88. After training, I ran it on the whole training set and saved result in "erroran-2.csv". 

```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
from PIL import Image
```

```python
compare_df = pd.read_csv("../input/errorfortrain/erroran-2.csv",index_col=[0])
compare_df
```

"label" is ground truth from "train.csv". "p_label" and "prob" is predicted by my model.

```python
error_df = compare_df[compare_df['label']!=compare_df['p_label']]
correct_df = compare_df[compare_df['label']==compare_df['p_label']]
error_df
```

Error rate is 0.12. Let's focus on "Healthy" class where my model disagrees with ground truth.

```python
error_4_x = error_df[error_df['label']==4]
error_4_x
```

```python
def show_image(image_ids, labels, plabels):
    plt.figure(figsize=(16, 12))
    for ind, (image_id, label, plabel) in enumerate(zip(image_ids, labels, plabels)):
        plt.subplot(3, 3, ind + 1)
        image = Image.open(os.path.join('../input/cassava-leaf-disease-classification/train_images',image_id))
        plt.imshow(image)
        plt.title(f"Class: {label} - Predict: {plabel}", fontsize=12)
        plt.axis("off")
    plt.show()
    
def show_batch(batch_index, df):
    image_ids = df[9*batch_index:9*(batch_index+1)]["image_id"].values
    labels = df[9*batch_index:9*(batch_index+1)]["label"].values
    plabels = df[9*batch_index:9*(batch_index+1)]["p_label"].values
    show_image(image_ids, labels, plabels)
    
def show_one_inbatch(batch_index, i,df):
    image_id = df['image_id'].values[9*batch_index+i]
    image = Image.open(os.path.join('../input/cassava-leaf-disease-classification/train_images',image_id))
    plt.figure(figsize=(10, 10))
    plt.imshow(image)
    plt.title(df[df['image_id']==image_id]['prob'].values[0], fontsize=18)
    plt.show()
```

```python
show_batch(0,error_4_x)
```

Most of them don't look healthy.

```python
show_one_inbatch(0, 0, error_4_x)
```

The first one should be classified as "Cassava Bacterial Blight(CBB)" as my model suggests. Let's look at ones wrongfully predicted as healthy.

```python
error_x_4 = error_df[error_df['p_label']==4]
show_batch(0,error_x_4)
```

```python
show_one_inbatch(0, 4, error_x_4)
```

Though predicted as healthy, the probability for CBB is 0.27, only 0.02 less than healthy. So I think maybe my model isn't that bad. Looking at how model actually performs does cheer me up. You should try it too. Thanks for your time.
