```python
import cv2
import json 

BASE_DIR = "../input/cassava-leaf-disease-classification/"
input_files = os.listdir(os.path.join(BASE_DIR, "train_images"))
train_df = pd.read_csv(os.path.join(BASE_DIR, "train.csv"))

# Read class mapping as dict
with open(os.path.join(BASE_DIR, "label_num_to_disease_map.json")) as file:
    map_classes = json.loads(file.read())

# Map clas to name
train_df ["class_name"] = train_df["label"].astype(str).map(map_classes)
```

```python
# Look at class label distribution
train_df['class_name'].value_counts().plot(kind="barh")
print(train_df['class_name'].value_counts())
```

```python
## The data is imbalanced and has a high number of CMD class examples
```

```python
import matplotlib.pyplot as plt

def visualize_batch(image_ids, labels):
    plt.figure(figsize=(20, 16))
    
    for ind, (image_id, label) in enumerate(zip(image_ids, labels)):
        plt.subplot(4, 4, ind + 1)
        image = cv2.imread(os.path.join(BASE_DIR, "train_images", image_id))
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        plt.imshow(image)
        plt.title(f"Class: {label}", fontsize=12)
        plt.axis("off")
    
    plt.show()
```

```python
# Sample df and visualize labels
iters = 3
num_samples = 16
for i in range(iters):
    tmp_df = train_df.sample(num_samples)
    image_ids = tmp_df["image_id"].values
    labels = tmp_df["class_name"].values

    visualize_batch(image_ids, labels)
```

```python
# Visualize unique classes
unique_df = train_df.groupby('class_name').aggregate({'image_id': max})
image_ids = unique_df["image_id"].values
labels = unique_df.index.values
visualize_batch(image_ids, labels)
```
