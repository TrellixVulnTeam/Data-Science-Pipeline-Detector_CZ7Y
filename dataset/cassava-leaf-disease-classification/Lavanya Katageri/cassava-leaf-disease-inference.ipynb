```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import os
ROOT_DIR = '../input/cassava-leaf-disease-classification/'
os.listdir(ROOT_DIR)
```

```python
from PIL import Image
from keras.models import load_model
```

# Load the best model from training

```python
TEST_DIR = '../input/cassava-leaf-disease-classification/test_images/'
```

```python
# load the best saved model as a new model
new_model = load_model('../input/cassava-leaf-disease-first-look-and-training/best_model.hdf5')
```

```python
new_model.summary()
```

# Prediction and submission

```python
IMG_SIZE = 300
size = (IMG_SIZE,IMG_SIZE)
```

```python
test_images = os.listdir(TEST_DIR)
preds = []
for image in test_images:
    img = Image.open(TEST_DIR + image)
    img = img.resize(size)
    img = np.expand_dims(img, axis=0)
    preds.extend(new_model.predict(img).argmax(axis = 1))
```

```python
preds
```

```python
sub = pd.DataFrame({'image_id': test_images, 'label': preds})
display(sub)
sub.to_csv('submission.csv', index = False)
```
