```python
#imports, setting the seed
import random
random.seed(42)
import pandas as pd
from fastai.vision.all import *
```

```python
# reading csv from input data
# Training with 10k input files for the trial submission
base_path="/kaggle/input/cassava-leaf-disease-classification"
df_train = pd.read_csv(base_path+"/train.csv")[:10000]
```

```python
#creating the data loader
dls = ImageDataLoaders.from_df(df_train, path=base_path+"/train_images", valid_pct=0.2,
                               seed=42, fn_col="image_id", label_col="label",bs=64,item_tfms=Resize(256))
```

```python
#training the model
learn = cnn_learner(dls, resnet18, metrics=error_rate, model_dir="/kaggle/working/")
learn.fine_tune(1)
```

```python
#saving the model
learn.export( Path("/kaggle/working/super_beginner_trained_model.pkl"))
```
