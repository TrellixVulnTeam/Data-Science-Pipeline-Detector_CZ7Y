* 512x512: https://www.kaggle.com/konradb/resized-data-512


```python
import os

import numpy as np 
import pandas as pd 
import openslide
import matplotlib.pyplot as plt
from PIL import Image
import cv2
from tqdm.notebook import tqdm
import skimage.io
from skimage.transform import resize, rescale
```

```python
newsize = 512
```

```python
data_dir = '/kaggle/input/cassava-leaf-disease-classification/train_images/'
save_dir = "/kaggle/train_images/"
os.makedirs(save_dir, exist_ok=True)
```

```python
train_labels = pd.read_csv('/kaggle/input/cassava-leaf-disease-classification/train.csv')
```

```python
for img_id in tqdm(train_labels.image_id):
    load_path = data_dir + img_id
    save_path = save_dir + img_id 
    
    image = cv2.imread(load_path)
    img = cv2.resize(image, (newsize, newsize))
    cv2.imwrite(save_path, img)
```
