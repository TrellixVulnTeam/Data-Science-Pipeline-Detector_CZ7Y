```python
def load_image(path , image_id):
    image = cv2.imread(path + image_id)
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB).astype(np.uint8)
```

```python
def norm_img(general_path , img_path, new_general_path):
    """
    norm img nim , max 0,1 for RGB images \
    
    
    
    """
    img = load_image(general_path , img_path)
    norm_image = cv2.normalize(img , None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX,
                               dtype=cv2.CV_32F)
    
    print(norm_image)
    cv2.imwrite(new_general_path +  img_path  , cv2.cvtColor(norm_image , cv2.COLOR_RGB2BGR))
    print(1)
```

```python
import zipfile
from tqdm import tqdm


def zip_and_remove(path):
    ziph = zipfile.ZipFile(f'{path}.zip', 'w', zipfile.ZIP_STORED)
    
    for root, dirs, files in os.walk(path):
        for file in files:
            file_path = os.path.join(root, file)
            ziph.write(file_path)
            os.remove(file_path)
    
    ziph.close()
```

```python
zip_and_remove(new_dir_name)
```
