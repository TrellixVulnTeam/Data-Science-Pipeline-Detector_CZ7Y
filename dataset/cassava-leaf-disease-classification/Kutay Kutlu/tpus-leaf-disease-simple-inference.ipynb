```python
# Importing useful libraries

import pandas as pd 
import numpy as np
import os

import keras
from keras.models import load_model
from PIL import Image
```

# Cassava Leaf Disease - TPU Tensorflow - Inference

### This is the inference part of the work, the training notebook can be found here 

https://www.kaggle.com/kutaykutlu/tpus-resnet50-leaf-disease

```python
model = load_model('../input/my-cassava-dataset/model_cassava.h5')
```

```python
img_size = 224
test_images = os.listdir('/kaggle/input/cassava-leaf-disease-classification/test_images/')
y_preds = []

for i in test_images:
    image = Image.open(f'/kaggle/input/cassava-leaf-disease-classification/test_images/{i}')
    image = image.resize((img_size, img_size))
    image = np.expand_dims(image, axis=0)
    y_preds.append(np.argmax(model.predict(image)))
```

```python
df_sub = pd.DataFrame({'image_id': test_images, 'label': y_preds})
display(df_sub)
```

```python
df_sub.to_csv('submission.csv',index=False)
```
