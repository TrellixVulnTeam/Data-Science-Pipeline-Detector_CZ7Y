# Browsing the Private Scores

```python
import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
from sklearn import datasets, linear_model
```

### The blue verticle lines indicate all the private scores in Private Leader Board.
### The red lines indicate the places of 0 and 1

```python
f = open("../input/cassava-lb/Private_LB.log",'r')
f = list(map(float,f.read().split()))
array = np.array(f)

plt.hlines(1,-0.05,1.05,color='k')
plt.xlim(-0.1,1.1)
plt.ylim(0.5,1.5)

y = np.ones(np.shape(array))
y2 = np.ones((2))
plt.plot(array,y,'|', ms=30)
plt.plot([0,1],y2,'r|', ms=60)
plt.axis('off')
plt.show()
```

# Let's draw the Public - Private graph

```python
df = pd.read_csv("../input/cassava-lb/Private_Public.log", names=["data"],encoding='utf-8')
df2 = pd.DataFrame(columns=['private', 'public'])
df.head(10)
```

### We submitted 204 times and only 187 submissions are not failed ones.

```python
for i in range(0,len(df),2):
    try:
        df2 = df2.append({'private':float(df["data"][i].replace('\u200b','')),'public':float(df["data"][i+1].replace('\u200b',''))},ignore_index=True)
    except:
        break
print(len(df2))
```

### Let's just sample the private scores above 0.8

```python
df2 = df2[df2['private']>0.8]
length = len(df2)
print(length)
```

```python
df2.head(10)
```

```python
private = df2['private'].to_numpy().reshape(length,1)
public = df2['public'].to_numpy().reshape(length,1)
```

```python
regr = linear_model.LinearRegression()
regr.fit(public, private)
```

```python
# plot it as in the example at http://scikit-learn.org/
plt.scatter(public, private,  color='black')
plt.plot(public, regr.predict(public), color='blue', linewidth=3)
plt.title("Public - Private")
plt.xlabel("Public")
plt.ylabel("Private")
plt.show()
print(regr.coef_)
```
