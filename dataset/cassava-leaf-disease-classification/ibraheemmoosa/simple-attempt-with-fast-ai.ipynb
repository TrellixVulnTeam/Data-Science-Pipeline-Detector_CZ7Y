```python
from fastai.data.all import *
from fastai.vision.all import *
```

```python
tr_img_path = Path('../input/cassava-leaf-disease-classification/train_images')
```

```python
tr_img_lbl_path = Path('../input/cassava-leaf-disease-classification/train.csv')
```

```python
tr_df = pd.read_csv(tr_img_lbl_path)
```

```python
tr_df['label'].value_counts()
```

```python
dls = ImageDataLoaders.from_df(tr_df, path=tr_img_path, seed=42, bs=8,# item_tfms=Resize(400, 300), 
                               batch_tfms=aug_transforms(flip_vert=True, batch=True, max_rotate=45.0))
```

```python
dls.show_batch()
```

```python
learn = cnn_learner(dls, resnet50, metrics=error_rate, path=Path('./'))
```

```python
learn.fine_tune(20)
```

```python
save_model('fast-ai-resnet-50.pth', learn, None, with_opt=False)
```

```python
learn.show_results()
```
