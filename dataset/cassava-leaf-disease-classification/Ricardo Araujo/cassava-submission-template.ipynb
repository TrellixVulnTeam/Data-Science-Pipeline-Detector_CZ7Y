```python
import numpy as np 
import pandas as pd

from fastai.vision.all import *
```

```python
df_train = pd.read_csv("../input/cassava-leaf-disease-classification/train.csv")
dls = ImageDataLoaders.from_df(df_train, path="../input/cassava-leaf-disease-classification/train_images", item_tfms=Resize(224))
```

```python
# Antes de rodar esta linha, é necessário fazer o upload do modelo treinado como se fosse um dataset
# Clicar em "+ Add data" nas opções da esquerda e fazer o upload do modelo previamente gravado (aqui chamado meuModelo.pth)

!cp ../input/modelo-teste/meuModelo.pth /tmp
```

```python
# Prepara o modelo e carregar

learn = cnn_learner(dls, alexnet, metrics=[error_rate,accuracy], pretrained=False, model_dir="/tmp/")
learn = learn.load("/tmp/meuModelo")
```

```python
# Aplicar o modelo no conjunto de teste

test_files = get_image_files('../input/cassava-leaf-disease-classification/test_images')
test_dl = learn.dls.test_dl(test_files, with_label=True)
pred_probas, x, pred_classes = learn.get_preds(dl=test_dl, with_decoded=True) 
```

```python
# Gravar os resultados (deve ter uma maneira mais eficiente...)

import os

rows = []
for file, label in zip(test_files, pred_classes):
    row = {'image_id': os.path.basename(file), 'label': int(label)}
    rows.append(row)
df = pd.DataFrame(rows)
df.to_csv('./submission.csv', index=False)
    
```
