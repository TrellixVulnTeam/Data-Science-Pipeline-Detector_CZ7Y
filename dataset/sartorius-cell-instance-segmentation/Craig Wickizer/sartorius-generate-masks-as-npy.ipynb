```python
import numpy as np
import pandas as pd

import os
```

#### Train csv contains all individual masks. Goal of notebook is to combine all annotations into a single np array for training

```python
df = pd.read_csv('../input/sartorius-cell-instance-segmentation/train.csv')
```

```python
df.describe()
```

#### All images are 520x704

```python
height = 520
width = 704
```

# Make all Masks

```python
#Go through all images
for file_id in df['id'].unique():
    file_name = f'{file_id}_mask.npy'
    test_mask = np.zeros((height*width))
    
    masks = df[df['id']==file_id]['annotation']
    
    #Making all masks value 1 in np array
    for mask in masks:
        pixel = []
        length = []
        for i, val in enumerate(mask.split()):
            if i % 2 == 0:
                pixel.append(int(val)-1)
            else:
                length.append(int(val))
        for pixel, length in zip(pixel,length):
            test_mask[pixel-1:pixel+length] = 1
    
    test_mask = test_mask.reshape((height,width))
    
    np.save(file_name,test_mask)
```
