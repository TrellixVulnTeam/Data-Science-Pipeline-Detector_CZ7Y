# Simple overlap-removing function

Hello community,

Below you can find a 3-line overlap-removing function. 
It takes two 2D numpy arrays as arguments and returns the first one (`mask`) with all the pixeles that overlapped with the second parameter (`other_mask`) removed.


It is just a modified version of the one used in the kernel [ðŸ¦  Sartorius - Starter Torch Mask R-CNN [LB=0.056]](https://www.kaggle.com/julian3833/sartorius-starter-torch-mask-r-cnn-lb-0-056).


Please **do upvote** if you find this useful.

```python
import numpy as np

def remove_overlapping_pixels(mask, other_mask):
    if np.sum(np.logical_and(mask, other_mask)) > 0:
        mask[np.logical_and(mask, other_mask)] = 0
    return mask
```

## Usage demonstration

```python
import matplotlib.pyplot as plt

m1 = np.zeros((520, 704))
m1[100:200, 100:200] = 1
plt.imshow(m1);
```

```python
m2 = np.zeros((520, 704))
m2[180:280, 180:280] = 1
plt.imshow(m2);
```

```python
plt.imshow(m1)
plt.imshow(m2, alpha=0.5)
plt.show()
```

```python
plt.imshow(m1[170:210, 170:210])
plt.imshow(m2[170:210, 170:210], alpha=0.5)
plt.show()
```

```python
m1_clean = remove_overlapping_pixels(m1, m2)
```

```python
plt.imshow(m1_clean);
```

```python
plt.imshow(m2);
```

```python
plt.imshow(m1_clean)
plt.imshow(m2, alpha=0.5)
plt.show();
```

```python
plt.imshow(m1_clean[170:210, 170:210])
plt.imshow(m2[170:210, 170:210], alpha=0.5)
plt.show();
```
