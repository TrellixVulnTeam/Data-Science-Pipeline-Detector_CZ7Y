# This Notebook contains code to programatically test whether the Internet is Enabled

For some reason it works fine within this competition (Sartorius -- Cell Instance Segmentation), but not as a free standing Notebook.

```python
from urllib.request import Request, urlopen

def is_internet_enabled(url='http://example.com/'):
    try:
        urlopen(Request(url))
        return True
    except:
        return False
    
is_internet_enabled()
```
