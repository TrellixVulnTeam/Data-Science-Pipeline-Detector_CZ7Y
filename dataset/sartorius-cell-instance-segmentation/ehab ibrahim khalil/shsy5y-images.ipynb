```python
import sklearn
import torchvision
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
from PIL import Image, ImageEnhance
import albumentations as A
from skimage import measure
import json
from matplotlib.path import Path
```

```python
train_img_path = '../input/sartorius-cell-instance-segmentation/train'
train_csv_path = '../input/sartorius-cell-instance-segmentation/train.csv'
shs_json_path = '../input/sartorius-cell-instance-segmentation/LIVECell_dataset_2021/annotations/LIVECell_single_cells/shsy5y/livecell_shsy5y_train.json'
shs_img_path = '../input/sartorius-cell-instance-segmentation/LIVECell_dataset_2021/images/livecell_train_val_images/SHSY5Y'
```

```python
train_df = pd.read_csv(train_csv_path)
train_df.head()
```

```python
cell_types   = {}
for id,row in train_df.groupby(['id']):
    cell_type = row.cell_type.unique()[0]
    if not cell_type in cell_types:
        cell_types[cell_type]=[]
    cell_types[cell_type].append(id)
 
stats = {}
for cell_type,ids in cell_types.items():
    stats[cell_type] = len(ids)
fig = plt.figure(figsize=(5,5))
ax  = fig.subplots()
ax.bar(stats.keys(),stats.values())
ax.set_title('Cell Types')
```
