# 3.DataFrame.items() => column name, column Series

### Iterate over (column name, Series) pairs.

```python
for label, content in train.items():
    print("label=>",label)
    print("content=>")
    print(content)
    print("")
```

# 4.DataFrame.iteritems()

### Iterate over (column name, Series) pairs.

dict.items() returns a list of 2-tuples ([(key, value), (key, value), ...]), 
whereas dict.iteritems() is a generator that yields 2-tuples. 

The former takes more space and time initially, 

but accessing each element is fast, whereas the second takes less space 
and time initially, but a bit more time in generating each element.

```python
for label,item in train.iteritems():
    print(label)
    print(item)
    print("")
    
```
