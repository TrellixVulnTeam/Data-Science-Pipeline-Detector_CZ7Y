# Submssion

The idea of this notbook is to quickly submit a privatly calculatet file

```python
def isNaN(num):
    return num != num

sample_submission = pd.read_csv('../input/landmark-recognition-2021/sample_submission.csv')

local_df = pd.read_csv('../input/glm-submission/submission.csv',index_col=0)


for i, row in sample_submission.iterrows():
    if row['id'] in local_df.index:
        if isNaN(local_df.loc[row['id']].landmarks): continue
        sample_submission.landmarks.loc[i] = local_df.loc[row['id']].landmarks

sample_submission.to_csv('submission.csv', index=False)
```
