```python
import xlrd
from pandas_profiling import ProfileReport
import sweetviz as sv
from autoviz.AutoViz_Class import AutoViz_Class


import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import scipy
import cv2
import tensorflow as tf
import PIL
import seaborn as sns
import os
```

```python
train_data = pd.read_csv('../input/landmark-recognition-2021/train.csv')
train_data.head()
```

```python
fig = plt.figure(figsize =(12, 10))

ax = fig.add_axes([0, 0, 1, 1])

bp = ax.boxplot(train_data['landmark_id'])

plt.show()
```

```python
train_data['landmark_id'].nunique()
```

## **Pairplot**

```python
plt.figure(figsize=(15, 12))
sns.pairplot(train_data)
plt.show()
```

## **Displot**

```python
plt.figure(figsize=(12, 12))
chart = sns.displot(train_data)
chart.set_xticklabels( rotation=90)
plt.show()
```

## **DistPlot**

```python
plt.figure(figsize=(12, 12))
chart = sns.distplot(train_data['landmark_id'])

plt.show()
```

## **Pandas Profiling**

```python
prof = ProfileReport(train_data)
prof.to_widgets()
```
