```python
import os


import random
import seaborn as sns
import cv2

# General packages
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import PIL
import IPython.display as ipd
import glob
import h5py
import plotly.graph_objs as go
import plotly.express as px
from PIL import Image
from tempfile import mktemp
from colorama import Fore, Back, Style

# Setting color palette.
plt.rcdefaults()
plt.style.use('dark_background')
```

* train.csv: only contains two columns
    * id: image id
    * landmark_id: target landmark id

```python
BASE_PATH = '../input/landmark-recognition-2021'

TRAIN_DIR = f'{BASE_PATH}/train'
TEST_DIR = f'{BASE_PATH}/test'


train = pd.read_csv(f'{BASE_PATH}/train.csv')
submission = pd.read_csv(f'{BASE_PATH}/sample_submission.csv')
train.shape
```

```python
landmark=train.groupby('landmark_id').count()
landmark_df=landmark
landmark_df['frequency']=landmark_df['id']
landmark_df=landmark_df.drop('id',axis=1)
```

**Unique number of landmarks in data set**

there are 81313 unique landmarks

```python
n=train['landmark_id'].nunique()   
sorted_df=landmark_df.sort_values(by='frequency',ascending=False).reset_index()
print(n)
```

# Number of images per landmark

* lets see top 50 landmarks with highest number of images in dataset

```python
import warnings
warnings.filterwarnings('ignore')

landmark_df=sorted_df.head(50)
landmark_df['landmark_id'] =  landmark_df.landmark_id.apply(lambda x: f'landmark_{x}')

fig = px.bar(landmark_df, y="frequency", x="landmark_id",color='landmark_id', orientation='v',
             hover_data=["landmark_id", "frequency"],
             height=1000,
             title='Number of images per landmark_id (Top 50 landmark_ids)')
fig.show()

```

**lets see last 10 landmarks with lowest number of images in dataset**

  * these landmarks only have 2 images 

```python
landmark_df1=sorted_df.tail(10)

landmark_df1['landmark_id'] =  landmark_df1.landmark_id.apply(lambda x: f'landmark_{x}')

fig = px.bar(landmark_df1, y="frequency", x="landmark_id",color='landmark_id', orientation='v',
             hover_data=["landmark_id", "frequency"],
             height=1000,
             title='Number of images per landmark_id (Top 50 landmark_ids)')
fig.show()
```

# landmark with higgest number of images 

  * id number 138982

```python
import PIL
from PIL import Image, ImageDraw


def display_images(images, title=None): 
    f, ax = plt.subplots(6,6, figsize=(18,22))
    if title:
        f.suptitle(title, fontsize = 30)

    for i, image_id in enumerate(images):
        image_path = os.path.join(TRAIN_DIR, f'{image_id[0]}/{image_id[1]}/{image_id[2]}/{image_id}.jpg')
        image = Image.open(image_path)
        
        ax[i//6, i%6].imshow(image) 
        image.close()       
        ax[i//6, i%6].axis('off')
    plt.show() 
  
samples = train[train.landmark_id == 138982].sample(30).id.values

display_images(samples)
```

```python
samples = train[train.landmark_id == 126637].sample(30).id.values

display_images(samples)
```
