```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_palette("Set2")
from wordcloud import WordCloud, STOPWORDS
```

```python
# load dataset
questions = pd.read_csv('../input/questions.csv')
answers = pd.read_csv('../input/answers.csv')
students = pd.read_csv("../input/students.csv")
professionals = pd.read_csv('../input/professionals.csv')
```

```python
students.head(2)
```

```python
professionals.head(2)
```

```python
answers.head(2)
```

```python
questions.head(2)
```

## Student's with particular questions 

```python
student_question = students.merge(right = questions, how = 'inner', left_on='students_id', right_on='questions_author_id')
```

```python
student_question.head(2).T
```

## Student's Questions with Answers

```python
sq_answers = student_question.merge(right=answers, how='inner', left_on='questions_id', right_on='answers_question_id')
```

```python
sq_answers.head(2).T
```

## Student's Questions answerd by Professionals 

```python
final_df = sq_answers.merge(right=professionals, how='inner', left_on='answers_author_id', right_on='professionals_id')
```

```python
final_df.head(2).T
```

```python
final_df.describe().T
```

## Where most of the student resides üè°

```python
final_df['students_location'].fillna("N/A", inplace=True)
```

```python
all_locations = ','.join(final_df['students_location'].values)

wordcloud = WordCloud(width=1500, height=800).generate(all_locations)

plt.figure(figsize=(20, 8))
# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

> __Most of students are from `New York`, `North Carolina`, `San Jose, California`, `Rhode Island`.__
