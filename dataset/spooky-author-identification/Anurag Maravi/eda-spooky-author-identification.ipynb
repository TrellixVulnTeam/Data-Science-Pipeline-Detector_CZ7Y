### Data
If you have not read the [data](https://www.kaggle.com/c/spooky-author-identification/data)  description, it is mentioned below.

* **id** - a unique identifier for each sentence
* **text** - some text written by one of the authors
* **author** - the author of the sentence

**EAP**: Edgar Allan Poe
**HPL**: HP Lovecraft
**MWS**: Mary Wollstonecraft Shelley

### Challenge
Build a model that predicts the author of excerpts from horror stories by Edgar Allan Poe, Mary Shelley, and HP Lovecraft.

In this notebook, we will explore the features given in the dataset.

Import the modules required

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import seaborn as sns
color = sns.color_palette()
%matplotlib inline
```

Load the training data

```python
train_df = pd.read_csv("../input/train.csv")
train_df.head()
```

```python
train_df.describe()
```

## Checking null values

```python
labels = []
values = []
for col in train_df.columns:
    labels.append(col)
    values.append(train_df[col].isnull().sum())
    print(col, values[-1])
```

As we can see that there is no missing data in this data set we can head to check the occurences of the attributes.

```python
cnt_author = train_df['author'].value_counts()
plt.figure(figsize=(15,10))
sns.barplot(cnt_author.index, cnt_author.values, alpha=0.8, color=color[0])
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xlabel('Author', fontsize=12)
plt.title('Count of occurence of author works', fontsize=15)
plt.xticks(rotation='vertical')
plt.show()
```

Clearly, the work by EAP has the maximum occurences in our training set followed by MWS and HPL with minimum occurence.

**EAP > MWS > HPL**

## Word cloud
