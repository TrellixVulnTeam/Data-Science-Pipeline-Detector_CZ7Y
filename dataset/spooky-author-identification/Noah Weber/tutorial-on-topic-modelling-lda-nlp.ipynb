**NOTE** Even tough we did use stopwords argument some of the words that do not have real predictive potential are to be found in the world cloud.  WE need better pre-processing----NLP

**NLP** Independent of the task at hand, wether we want to predict authors, find the underlying topics or just cluster the words. Computer can not process textual data. WE need to make it computer readable. Underneath are few crucial pre-processing steps to convert the input raw text into a form that is readable by your model and the machine. In the end we can summarize text pre-processing in the following points

* **Tokenization** - Splitting the text into smaller blocks (words-USUALLY!) Exactly what we did above with str.split but now we have a library that is far more efficient 
* **Stopwords**  - Neglecting the words that do not have any predictive power. That includes frequent and very infrequent words!
*  **Stemming** - Best understood on the example. Sleep, Sleepy, Sleeping. They are all describing, revolving around the same subject. Why count them all. Just extract the "root"-"Stem" of the word, in this case Sleep and work with that.
* **Vectorization** - Simplest is the famous bag-of-words approach, where you create a matrix (for each document or text in the corpus). In the simplest form, this matrix stores word frequencies (word counts) and is oft referred to as vectorization of the raw text. Underneath it should be more clear wit han example. But keep one thing in mind. This is a final step where everything turns into numerical variables. (For more depth I wrote a paper regarding NLP that goes into more complex weighing techniques---visit my LinkedIN account!)

**Turning words into numbers - Vectorization** Now there are is a number of techniques that can be used to make this transition. WE are going to be talking about bag of words approach. (Name comes from the fact that we just count appearances and nothing else, i.e. just drawing words from the bag and counting) Steps are as follows:
*  Count the words that are to be found in the entire text, build our dictionary in a sense.
* Record the occurance of that specific word. For every element of the list we will be making a vector that checks the occurance
Best way is to look at an Example: (NOTE that CountVectorizer is from sklearn library)


```python
sentence = ["I study Math", 
            "I enjoy doing Math"]
vectorizer = CountVectorizer(min_df=0)
sentence_transform = vectorizer.fit_transform(sentence)
```

Cool! Now we are ready to start modeling. But in the beginning I mentioned that we can pipe all of the pre-processing steps together to make it more efficient. For that we will use even more powerful library sklearn, where Sklearn's tokenizer discards all single character terms like ('a', 'w' etc) and also lower cases all terms by default unlike the NLTK tokenizer.


When you vectorize the raw text with CountVectorizer, the dual stages of tokenizing and stopwords filtering are automatically included as a high-level component. 


Unfortunately, there is no built-in lemmatizer in the vectorizer so we are left with a couple of options. Either implementing it separately everytime before feeding the data for vectorizing or somehow extend the sklearn implementation to include this functionality. We will just extend the class (concepts of object oriented programming) with the lemmatizing functionality that we need, i.e. extend the CountVectorizer class by overwriting the "build_analyzer" method as follows:

**Extending**

```python
lemm = WordNetLemmatizer()
class LemmaCountVectorizer(CountVectorizer):
    def build_analyzer(self):
        analyzer = super(LemmaCountVectorizer, self).build_analyzer()
        return lambda doc: (lemm.lemmatize(w) for w in analyzer(doc))
```

**Applying it on text**

```python
# Storing the entire training text in a list
text = list(train.text.values)
# Calling our overwritten Count vectorizer
tf_vectorizer = LemmaCountVectorizer(max_df=0.95, 
                                     min_df=2,
                                     stop_words='english',
                                     decode_error='ignore')
tf = tf_vectorizer.fit_transform(text)
print(tf_vectorizer.get_feature_names()) #complete Dictionary of our words
```

**LDA** Modelling. Finally we are ready to fit the model. There are many parameters that need to be tuned for LDA. As always in order to do it properly fit  it we need to have some deeper understanding. Since this tutorial is concerned with implementation, we are going to state hwo we got to the chosen parameters but the reader should explore the Algorithm a bit deeper on tis own. Implementation of LDA can be done with sklearn or gensim. We are going to sitck with skicit learn library. Great source is [http://videolectures.net/mlss09uk_blei_tm/](http://) video lecture at Princeton, also offical documentation [http://scikit-learn.org/stable/modules/generated/sklearn.decomposition.LatentDirichletAllocation.html](http://). Also to understand what hyper-parameters should be choosen we can consult some of the novel approaches [https://www.ncbi.nlm.nih.gov/pmc/articles/PMC4597325/](http://) or use  KMeans + Latent Semantic Analysis Scheme to find out n_components (number of topics) whereby the number of Kmeans clusters and number of LSA dimensions were iterated through and the best silhouette mean score. Let us take the "cooked" values.

```python
lda = LatentDirichletAllocation(n_components=11, max_iter=5,
                                learning_method = 'online',
                                learning_offset = 50.,
                                random_state = 0)

lda.fit(tf)
lda.components_
```

So what happend? how can we now extract the desired Info/Topics?
First we are going to define helper function that is going to print us top words.
This function takes the model and extracts the components of the applied model, along with the desired number of top words. Why do we need to build such a function in a first place? When fitting the lda, we can acces the topics created (in our case 11 ) with simple lda.components_   command (careful we are going to get numerical values when executing it). There we will have a lot of words in a first topc that are simply not representative. Hence let us look only at top 50 words. (Per topic)


```python
# Define helper function to print top words
def print_top_words(model, feature_names, n_top_words):
    for index, topic in enumerate(model.components_):
        message = "\nTopic #{}:".format(index)
        message += " ".join([feature_names[i] for i in topic.argsort()[:-n_top_words - 1 :-1]])
        print(message)
        print("="*70)
```

```python
nubmer_of_words=50
print("\nTopics in LDA model: ")
tf_feature_names = tf_vectorizer.get_feature_names() #note that tf_vectorizer is an LemmaCountVectorizer object and with this command we get the whole dictionary of words
print_top_words(lda, tf_feature_names, nubmer_of_words)
```

```python
first_topic = lda.components_[0]
second_topic = lda.components_[1]
third_topic = lda.components_[2]
fourth_topic = lda.components_[3]
```
