```python
def set_lower(data_list):
    df = data_list[0].copy()
    target = data_list[1]
    df[target] = df[target].apply(lambda x: " ".join(x.lower() for x in x.split()))
    return df
```

```python
def set_nonpunct(data_list):
    df = data_list[0].copy()
    target = data_list[1]
    df[target] = df[target].str.replace('[^\w\s]','')
    return df
```

```python
def set_nonstop_words(data_list):
    df = data_list[0].copy()
    target = data_list[1]
    from nltk.corpus import stopwords
    stop = stopwords.words('english')
    df[target] = df[target].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
    return df
```

```python
def set_lammatize(data_list):
    from textblob import Word
    df = data_list[0].copy()
    target = data_list[1]
    df[target] = df[target].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
    return df
```

```python
#Import library

def set_stemming(data_list):
    from nltk.stem import PorterStemmer
    df = data_list[0].copy()
    target = data_list[1]
    st = PorterStemmer()
    df[target].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))
    return df
```

```python
df = pd.read_csv("/kaggle/input/spooky-author-identification/train.zip")
print(df.head())
print(df['author'].unique())
```

```python
df = set_lower([df, 'text'])
df = set_nonpunct([df, 'text'])
df = set_nonstop_words([df, 'text'])
df = set_lammatize([df, 'text'])
```

```python
print(df)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
vectorizer.fit(df['text'])
vector = vectorizer.transform(df['text'])
print(vector.toarray())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(ngram_range=(2,2))
vectorizer.fit(df['text'])
vector = vectorizer.transform(df['text'])
print(vector.toarray())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
vectorizer.fit(df['text'])
print(vectorizer.idf_)
```

```python
import gensim
from gensim.models import Word2Vec
from sklearn.decomposition import PCA
from matplotlib import pyplot
from gensim.test.utils import common_texts

model = Word2Vec(sentences=common_texts, vector_size=100, window=5, min_count=1, workers=4)
model.train([["hello", "world"]], total_examples=1, epochs=1)
vector = model.wv['computer']
sims = model.wv.most_similar('computer', topn=10) 
print(vector)
print(sims)
```
