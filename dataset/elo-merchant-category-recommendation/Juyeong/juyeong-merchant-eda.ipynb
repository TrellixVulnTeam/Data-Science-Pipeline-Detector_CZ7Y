위의 두 변수들에 비해 특별한 이상치는 존재하지 않음.

지난 n개월 동안 활성화 되어있었던 개월 수를 나타내는 것이므로 중요한 것은 3, 6, 12개월에 해당하지 않는 merchant들.

이 변수를 이용하여 merchant의 성실성을 대용하는 변수를 생성할 수 있을 것으로 보인다.
ex) 3개월내에 1번이상, 12개월에 3개월 이상 영업을 쉬었다면(성실도가 낮다면, categorical 변수로 적용하는 것이 좋지 않을까?), 이는 해당 merchant와 주로 거래하는 고객의 로열티가 떨어질 수 있는 이유가 될 것이다.

```python
x1 = data.loc[data['active_months_lag3']!=3]
x2 = data.loc[data['active_months_lag6']!=6]
x3 = data.loc[data['active_months_lag12']!=12]

fig, ax = plt.subplots(1, 3, figsize = (16, 6));
x1['active_months_lag6'].value_counts().sort_index().plot(kind='bar', ax=ax[0], color='teal', title='3_Month');
x2['active_months_lag6'].value_counts().sort_index().plot(kind='bar', ax=ax[1], color='brown', title='6_Month');
x3['active_months_lag12'].value_counts().sort_index().plot(kind='bar', ax=ax[2], color='blue', title='12_Month');
```

N개월을 제외하고 나타낸 분포. 거의 선형적인 개형을보이고 있다. 

Merchant 의 성실성(active months) / 최근 성장도(sales/purchases) / 변수들을 생성하여 고객이 주로 이용했던 merchant가 긍정적인지 부정적인지를 이용하면 고객 로열티 예측에 충분히 설명력을 가질 수 있을 것으로 보인다.
