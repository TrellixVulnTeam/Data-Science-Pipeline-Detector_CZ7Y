```python
from autoviml.Auto_ViML import Auto_ViML
```

# We are going to try Auto_ViML first and see whether it does well

```python
target = "target"
```

```python
m, feats, trainm, testm = Auto_ViML(train, target, test,
                            sample_submission='',
                            scoring_parameter='', KMeans_Featurizer=False,
                            hyper_param='RS',feature_reduction=True,
                             Boosting_Flag='CatBoost', Binning_Flag=False,
                            Add_Poly=0, Stacking_Flag=False,
                                    Imbalanced_Flag=False,
                            verbose=1)
```

```python
testm.head(2)
```

```python
preds = testm[target+'_predictions']
preds[:5]
```

```python
sample_submission = pd.read_csv('/kaggle/input/elo-merchant-category-recommendation/sample_submission.csv')
```

```python
sample_submission['target'] = preds
sample_submission.to_csv('submission.csv', index=False)
```
