# Importing libs
I will be using RandomForestRegressor.

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
```

# Loading data
* Making x's and y's with the training data
* Loading from memory

```python
df = pd.read_csv('/kaggle/input/tabular-playground-series-jan-2021/train.csv')
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:, -1]
```

```python
x.head()
```

# Intializing the model
* N_Estimators--> number of search trees
* n_jobs-> to use all the avialable thread

```python
forest = RandomForestRegressor(n_estimators=500, n_jobs=-1)
```

# Fitting
Training phase on the data directly

```python
forest.fit(x, y)
```

# Test results
Getting results and submitting.

```python
test = pd.read_csv('/kaggle/input/tabular-playground-series-jan-2021/test.csv')
```

```python
output = pd.DataFrame(forest.predict(test),columns=['target'])
```

```python
output['id'] = test['id']
```

```python
ouput = output.set_index('id')
```

```python
output = ouput.reset_index()
```

```python
output.to_csv('submission1.csv', index=False)
```
