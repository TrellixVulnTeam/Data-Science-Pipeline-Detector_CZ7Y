## Data Load

```python
PATH = '/kaggle/input/tabular-playground-series-jan-2021/'
```

```python
import pandas as pd

train = pd.read_csv(PATH+'train.csv')
test = pd.read_csv(PATH+'test.csv')
submission = pd.read_csv(PATH+'sample_submission.csv')
```

```python
train.head()
```

```python
from pycaret.regression import *
```

## Settup data

```python
reg = setup(data=train, target='target', silent=True, session_id=2021)
```

## Check all models and choose 3 best ones

```python
best_3 = compare_models(include = ['xgboost', 'lightgbm', 'catboost'], sort='RMSE', n_select=3)
```

## Model Ensemble

```python
blended = blend_models(best_3, fold=5)
```

## Prediction

```python
pred_holdout = predict_model(blended)
```

## Re-train the model with whole data

```python
final_model = finalize_model(blended)
```

## Predict on test set

```python
predictions = predict_model(final_model, data=test)
```

```python
predictions
```

## Submission

```python
submission['target'] = predictions['Label']
```
