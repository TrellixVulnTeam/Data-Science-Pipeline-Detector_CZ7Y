```python
#import and assign to variables
train = pd.read_csv('../input/tabular-playground-series-jan-2021/train.csv')
test = pd.read_csv('../input/tabular-playground-series-jan-2021/test.csv')
sample_submission = pd.read_csv('../input/tabular-playground-series-jan-2021/sample_submission.csv')
```

```python
data = pd.concat([train, test], sort=False)
```

```python
data.head()
```

```python
data.drop(['id'], axis=1,inplace=True)
```

```python
train = data[:len(train)]
test = data[len(train):]
```

```python
y_train = train['target']
X_train = train.drop('target', axis = 1)
X_test = test.drop('target', axis = 1)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
from sklearn.linear_model import LinearRegression

clf = LinearRegression()
clf.fit(X_train, y_train)
```

```python
for i in range(14):
    print('coefficient'+ str(i+1) + '= ', clf.coef_[i]) #Output the coefficient of the explanatory variable
print('intercept = ', clf.intercept_) #Output intercept
```

```python
y_pred = clf.predict(X_test) #Predict the objective variable
```

```python
y_pred[:20]
```

```python
from sklearn.model_selection import cross_validate

score = cross_validate(clf,X_train,y_train,cv=5,n_jobs=-1)
print('fit_time',score['fit_time'])
print('score_time',score['score_time'])
print('test_score',score['test_score'])
```

```python
#submission
sub = sample_submission
sub['target'] = list(map(float, y_pred))
sub.to_csv("submission.csv", index=False)
```
