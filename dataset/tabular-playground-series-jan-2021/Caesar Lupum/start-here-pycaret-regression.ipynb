```python
from pycaret.regression import *
```

```python
train_data = pd.read_csv('../input/tabular-playground-series-jan-2021/train.csv')
test_data  = pd.read_csv('../input/tabular-playground-series-jan-2021/test.csv')
```

```python

features = ['cont1', 'cont2', 'cont3', 'cont4', 'cont5', 'cont6', 'cont7',
       'cont8', 'cont9', 'cont10', 'cont11', 'cont12', 'cont13', 'cont14']

X_train = train_data[features]
y_train = train_data["target"]
final_X_test = test_data[features]
```

```python
# Setup PyCaret
setup(data=train_data,          # Data used for training
      target='target',     # Target column name 
      fold=3,              # Number of fold tu use for cross-validation
      silent=True,         # No data type validation by user
      verbose=False,       # No print of the setup params grid
      normalize = True,
      session_id=2021);    # Seed
```

### Compare different regression models

```python
compare_models()
```

### let's do Lightgbm

```python
model = create_model('lightgbm')
```

##### Tune model hyper-parameters

```python
# Tune model
tuned_model = tune_model(model,                               # Model to tune
                         custom_grid=None,                    # Use default search space
                         optimize='RMSE',                     # Metrics to optimize
                         search_library='scikit-optimize',    # Underlying framework to use
                         n_iter=100,                           # Number max of iterations
                         choose_better=True)                  # Return best performing model
```

#### SHAP Values (impact on model output)

```python
interpret_model(tuned_model)
```

```python
# Train on all training data
final_model = finalize_model(tuned_model) 
```

#### Submission file

```python
prediction = predict_model(final_model, test_data)
prediction['Label'].rename('target').to_csv('submission.csv')
prediction.head()
```
