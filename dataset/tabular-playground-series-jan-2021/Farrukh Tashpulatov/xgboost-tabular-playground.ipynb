```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df_train = pd.read_csv('/kaggle/input/tabular-playground-series-jan-2021/train.csv')
df_test = pd.read_csv('/kaggle/input/tabular-playground-series-jan-2021/test.csv')
df_train.describe()
```

```python
X = df_train.drop(['target','id'], axis = 1)
y = df_train['target']
test_id = df_test.id
X_test = df_test.drop('id', axis = 1)
print(len(X), len(y))
```

```python
for column in X.columns:
#     plt.subplot(7, 2, i)
#     i += 1
    plt.hist(X[column])
    plt.title(column)
    plt.show()
    plt.close()
plt.show()
plt.close('all')
plt.hist(y)
```

```python
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
# X = X.drop(['cont5','cont14'], axis = 1)
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = .2, random_state = 1)
# y_train.head
model = XGBRegressor(n_estimators=100)
model.fit(X_train, y_train)
# model.score(X_val, y_val)
y_pred = model.predict(X_val)
mse = mean_squared_error(y_pred, y_val)
mse
```

```python
predictions = model.predict(X_test)
```

```python
output = pd.DataFrame({'id': test_id, 'target': predictions})
output.to_csv('my_submission.csv', index=False)
print("Your submission was successfully saved!")
```
