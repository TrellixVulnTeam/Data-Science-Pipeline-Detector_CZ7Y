#### 1.c Outliar Addressal

```python
# before proceeding further, lets first create a copy of the data-set.
# we will use simple imputer and strategy as mean for addressing outliars
```

```python
t2data = t1data.copy()
t2data_test=t1data_test.copy()
```

```python
def outliar_detection(col):
    Q1=t2data[col].quantile(0.25)
    Q3=t2data[col].quantile(0.75)
    IQR=Q3-Q1
    Lower_Whisker = Q1-1.5*IQR
    Upper_Whisker = Q3+1.5*IQR
    t2data[col][t2data[col]> Upper_Whisker] = np.nan
    t2data[col][t2data[col]< Lower_Whisker] = np.nan
    return t2data[col][t2data[col].isnull()]
```

```python
ol_columns=['cont1','cont2','cont3','cont4','cont5','cont6','cont7','cont8','cont9','cont10','cont11','cont12','cont13','cont14']
ol_columns=np.array(ol_columns)
```

```python
for i in ol_columns:
    outliar_detection(i)
```

```python
t2data.info()
```

```python
# Imputing the missing values with mean
columns=t2data.columns
imp_median = SimpleImputer(missing_values=np.nan, strategy='mean')

imp_median.fit_transform(t2data)
# imp_median.fit(t2data)
t2data=pd.DataFrame(imp_median.transform(t2data))
t2data.columns=columns
```

```python
def outliar_detection_test(col):
    Q1=t2data_test[col].quantile(0.25)
    Q3=t2data_test[col].quantile(0.75)
    IQR=Q3-Q1
    Lower_Whisker = Q1-1.5*IQR
    Upper_Whisker = Q3+1.5*IQR
    t2data_test[col][t2data_test[col]> Upper_Whisker] = np.nan
    t2data_test[col][t2data_test[col]< Lower_Whisker] = np.nan
    return t2data_test[col][t2data_test[col].isnull()]
```

#### Stacking

```python
estimators = [
    GradientBoostingRegressor(learning_rate= 0.008,max_depth= 6,n_estimators= 2000,random_state= 1,subsample= .8
                                  ,max_features='log2',min_samples_split=2000,min_samples_leaf = 50
                                 ),
    LGBMRegressor(n_estimators=300),
    XGBRegressor(eta= 0.008,min_child_weight=4,eval_metric='logloss',gamma=0.0,reg_alpha=0.005,max_depth= 6,n_estimators= 2000,random_state= 1,subsample= 1),
    CatBoostRegressor(n_estimators=300),
    RandomForestRegressor(n_estimators=200)
]
```

```python
S_train, S_test = stacking(estimators,                   
                           X, y, X_test,   
                           regression=True,
                           metric=mean_squared_error, 
                           n_folds=5,
                           shuffle=True,  
                           random_state=0,
                           verbose=2) 
```

```python
model = GradientBoostingRegressor(learning_rate= 0.008,max_depth= 6,n_estimators= 2000,random_state= 1,subsample= .8
                                  ,max_features='log2',min_samples_split=2000,min_samples_leaf = 50
                                 )
model=model.fit(S_train, y)
```

```python
y_pred = model.predict(S_test)
```
