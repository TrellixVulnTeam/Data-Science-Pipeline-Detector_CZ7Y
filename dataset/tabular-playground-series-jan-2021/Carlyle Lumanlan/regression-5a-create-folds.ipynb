```python
from sklearn import model_selection
```

```python
train_raw = pd.read_csv('/kaggle/input/tabular-playground-series-jan-2021/train.csv')
print(train_raw.head())
train_raw.shape
```

```python
train_raw['kfold'] = -1
```

```python
# create a Kfold split

kf = model_selection.KFold(n_splits=5, shuffle=True, random_state=34)

for fold, (train_indices, valid_indices) in enumerate(kf.split(X=train_raw)):
    train_raw.loc[valid_indices, 'kfold'] = fold
```

```python
train_raw.to_csv('train_raw_folds.csv', index=False)
```
