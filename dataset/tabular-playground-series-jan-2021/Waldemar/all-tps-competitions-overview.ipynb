### Read input csv's and extract metadata

```python
# !tree ../input/
base_path = Path('/kaggle/input/')
competition_dirs = [competition_dir for competition_dir in base_path.iterdir() 
                    if competition_dir.is_dir() and competition_dir.name.startswith('tabular-playground-series')]

competition_dirs.sort(key=lambda dir_name: datetime.datetime.strptime(dir_name.name[-8:], '%b-%Y').strftime('%Y-%m'), reverse=True) 
competition_dirs
```

```python
def get_df_metadata(comp_dir, nrows=None):
    
    def extract(prefix, csv_path, separator=' - '):
        metadata[f'{prefix}{separator}file (MB)'] = round(csv_path.stat().st_size / 2**20)
        df = pd.read_csv(csv_path, nrows=nrows)
        metadata[f'{prefix}{separator}in mem (MB)'] = round(df.memory_usage(deep=True).sum()/ 2**20 )
        
        metadata[f'{prefix}{separator}rows'] = df.shape[0]
        metadata[f'{prefix}{separator}columns'] = df.shape[1]
        metadata[f'{prefix}{separator}NaN count'] = df.isna().sum().sum()

    
    metadata = {}
    
    metadata['Name'] = comp_dir.name
    train_csv_path = comp_dir.joinpath('train.csv')
    
    if train_csv_path.exists():
        extract('Train', train_csv_path)
        
        test_csv_path = comp_dir.joinpath('test.csv')
        extract('Test', test_csv_path)
    else:
        data_csv_path = comp_dir.joinpath('data.csv')
        extract('Train', data_csv_path)
        
    return metadata


competitions_metadata = []

for directory in competition_dirs:
#     competitions_metadata.append(get_df_metadata(directory, np.random.randint(100,10000)))
    competitions_metadata.append(get_df_metadata(directory))
    clear_output(wait=True)
    pd.DataFrame(competitions_metadata)

df_metadata = pd.DataFrame(competitions_metadata)
```

```python
df = pd.merge(df_metadata, df_manual, on=['Name'])

base_url = 'https://www.kaggle.com/competitions/'

def get_yyyy_mm_from_name(comp_name: str) -> str:
    return datetime.datetime.strptime(comp_name[-8:], '%b-%Y').strftime('%Y-%m')
    
df['Name'] = df['Name'].apply(lambda x: f'<a href="{base_url}{x}" >{get_yyyy_mm_from_name(x)}</a>')

column_multi_idx = []
for c in df.columns.to_list():
    if ' - ' in c:
        column_multi_idx.append(c.split(' - ', maxsplit=2))
    else:
        column_multi_idx.append((c, ''))
df.columns = pd.MultiIndex.from_tuples(column_multi_idx)


HTML(df.to_html(escape=False))
```

# EDA

```python
# https://matplotlib.org/3.5.0/tutorials/colors/colormaps.html
col_map = plt.get_cmap('tab20')

fig, axs =plt.subplots(1, 4, figsize=(15, 10), layout='constrained')
fig.patch.set_facecolor('#E0E0E0')

_ = fig.suptitle('Train data - memory usage and data size', fontsize=20)

data = pd.DataFrame()
data['name'] = df['Name'].str[-11:-4]
data['File size in MB'] = df[('Train', 'file (MB)')]
data['df size in MB'] = df[('Train', 'in mem (MB)')]
data['Number of rows'] = df[('Train', 'rows')]
data['Number of columns'] = df[('Train', 'columns')]


data.sort_values(by='File size in MB', inplace=True)

def plt_barh(ax, width_col):
    _ = ax.xaxis.grid(color='gray', alpha=0.5)
    bars = ax.barh(y=data['name'], width=data[width_col], color=col_map.colors, label=width_col)
    _ = ax.bar_label(bars)
    old_lim = ax.get_xlim()
    new_lim = (old_lim[0], old_lim[1]*1.1)
    _ = ax.set_xlim(new_lim)
    _ = ax.set_title(width_col)
    _ = ax.set_facecolor('#CCC')
    
plt_barh(axs[0], 'File size in MB')
plt_barh(axs[1], 'df size in MB')
plt_barh(axs[2], 'Number of rows')
plt_barh(axs[3], 'Number of columns')
```
