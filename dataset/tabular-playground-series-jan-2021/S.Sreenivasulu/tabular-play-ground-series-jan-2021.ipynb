## Splitting the traindf into train and test with sklearn

```python
from sklearn.model_selection import train_test_split
```

```python
# Remove id, target columns 
X = traindf.drop(["id", "target"], axis = 'columns', inplace = False).values
y = traindf['target'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.30, random_state=85)
#Making testdf data as an array of values
test_submit = testdf.drop("id", axis=1).values
```

## Applying simple linear regression Model with SKlearn

```python
from sklearn.linear_model import LinearRegression 
regr_model = LinearRegression()
regr_model.fit(X_train, y_train)
```

### Predicting the results using the regression model fitted

```python
y_pred = regr_model.predict(X_test) 
```

## Linear Regression Results

### 1. mean_absolute_error

```python
 from sklearn.metrics import mean_absolute_error
 print(mean_absolute_error(y_test, y_pred))
```

### 2. Mean squared error regression loss

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred)
```

### 3. r2_score

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)  
```
