```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
training_set=pd.read_csv('../input/tabular-playground-series-jan-2021/train.csv')
training_set.set_index('id',inplace=True)
#print(training_set.head())
y=training_set['target']
y_train=training_set.iloc[:,:14]
#print(y_train.head())
#print(y.head())
```

```python
test_set=pd.read_csv('../input/tabular-playground-series-jan-2021/test.csv')
test_set.set_index('id',inplace=True)
print(test_set.head())
#../input/tabular-playground-series-jan-2021/test.csvy=test_set['target']
x_test=test_set.iloc[:,:14]
#print(y_train.head())
#print(y.head())
```

```python
col=training_set.columns
count=0
fig,axes=plt.subplots(7,2,figsize=(50,50))
for i in range(7):
    for j in range(2):
        axes[i][j].scatter(x=training_set['target'],y=training_set[col[count]])
        count+=1
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
linreg=LinearRegression().fit(y_train,y)
print(linreg.predict(x_test))
solution=pd.DataFrame({'id':x_test.index,"Target":linreg.predict(x_test)})
print(solution.head())
```
