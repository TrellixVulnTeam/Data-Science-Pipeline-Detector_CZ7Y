```python
train = pd.read_csv('../input/tabular-playground-series-jan-2021/train.csv',index_col='id')
train.head()
```

```python
test = pd.read_csv('../input/tabular-playground-series-jan-2021/test.csv',index_col='id')
test.head()
```

```python
print(train.shape)
print(test.shape)
```

```python
train.columns
```

```python
train.dtypes
```

```python
train.dropna(axis=0,inplace=True)
train.shape
```

```python
train.describe()
```

```python
y_train = train['target']
y_train.head()
```

```python
X_train = train.iloc[:,:-1]
X_train.head()
```

```python
df = test.index
df
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler,PolynomialFeatures
from xgboost import XGBRegressor
```

```python
X_train,X_valid,y_train,y_valid = train_test_split(X_train,y_train,test_size=0.1)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_valid.shape)
print(y_valid.shape)
```

```python
Input = [('scale',StandardScaler()),('model',XGBRegressor(n_estimators=50))]
```

```python
model = Pipeline(Input)
```

```python
model.fit(X_train,y_train)
```

```python
pred = model.predict(X_valid)
print(pred.shape)
```

```python
plt.figure(figsize=(8,8))
plt.title("Predicted Values")
sns.scatterplot(X_valid.index,pred)
```

```python
plt.figure(figsize=(8,8))
plt.title("Actual Values")
sns.scatterplot(X_valid.index,y_valid)
```

```python
rmse = mean_squared_error(pred,y_valid,squared=False)
```

```python
prediction = model.predict(test)
prediction
```

```python
output = pd.DataFrame({'id':df,'target':prediction})
output.to_csv("submission.csv",index=False)
```
