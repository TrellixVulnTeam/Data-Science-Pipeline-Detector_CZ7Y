# import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import seaborn as sns
from xgboost import XGBRegressor
from warnings import filterwarnings 
import xgboost as xgb
filterwarnings('ignore')
```

# getting data

```python
train = pd.read_csv('/kaggle/input/tabular-playground-series-jan-2021/train.csv',index_col  = 'id')
print('shape : {} '.format(train.shape))
train.head()
```

# correlation

```python
plt.figure(figsize = (14,6))
sns.heatmap(train.corr(),annot = True)
```

# visualization

```python
plt.figure(figsize = (14,16))
for i in range(len(train.columns)):
    plt.subplot(5,3,i+1)   
    sns.boxplot(train[train.columns[i]])
plt.show()
```

```python
plt.figure(figsize = (16,16))
for i in range(len(train.columns)):
    plt.subplot(5,3,i+1)   
    sns.distplot(train[train.columns[i]])
plt.show()
```

```python
plt.figure(figsize = (14,16))
for i in range(len(train.columns)):
    plt.subplot(5,3,i+1)   
    sns.scatterplot(x = train.columns[i],y = 'target',data = train)
plt.show()
```

# train-test split

```python
x = train.drop('target',axis=1)
y = train['target']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25,random_state = 42)
```

# model

```python
regressor = xgb.XGBRegressor(
                 colsample_bytree=0.5,
                 alpha=0.01563,
                 #gamma=0.0,
                 learning_rate=0.01,
                 max_depth=15,
                 min_child_weight=257,
                 n_estimators=4000,                                                                  
                 #reg_alpha=0.9,
                 reg_lambda=0.003,
                 subsample=0.7,
                 random_state=2020,
                 metric_period=100,
                 silent=1)

regressor.fit(x_train, y_train, early_stopping_rounds=6, eval_set=[(x_test, y_test)], verbose=1)
```

```python
df_test = pd.read_csv('/kaggle/input/tabular-playground-series-jan-2021/test.csv',index_col = 'id')
```

```python
df_tahmin = pd.DataFrame(regressor.predict(df_test),index= df_test.index,columns = ['target'])
```
