```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn import preprocessing

import xgboost as xgb
```

```python
train_df = pd.read_csv('/kaggle/input/tabular-playground-series-jan-2021/train.csv')
test_df = pd.read_csv('/kaggle/input/tabular-playground-series-jan-2021/test.csv')
```

```python
train = train_df.drop(['id','target'],axis=1)
train_target = train_df['target']
```

```python
X_train,X_test,y_train,y_test = train_test_split(train,train_target,test_size=0.2,random_state=0)
```

```python
regressor = xgb.XGBRegressor(
                 colsample_bytree=0.5,
                 alpha=0.03,
                 #gamma=0.0,
                 learning_rate=0.01,
                 max_depth=15,
                 min_child_weight=200,
                 n_estimators=4000,                                                                  
                 #reg_alpha=0.9,
                 reg_lambda=0.003,
                 subsample=0.7,
                 random_state=2021,
                 metric_period=100,
                 silent=1)

regressor.fit(X_train, y_train, early_stopping_rounds=6, eval_set=[(X_test, y_test)], verbose=1)
```

```python
predictions = regressor.predict(X_test)
print('mean square error on training data (vanilla model):',
      np.round(mean_squared_error(y_test, predictions, squared=False),5)) 
```

```python
test = test_df.drop('id', axis=1)
pred = regressor.predict(test)

submission = pd.DataFrame({ 'id': test_df['id'],
                            'target': pred})

print(submission.shape)

submission.to_csv("submission.csv", index=False)
```
