```python
y=df1['target'].values
```

```python
x=df1.drop(columns='target').values
```

```python
x
```

```python
#splitting the data set and ,making models to get the bet result
```

```python
#linear regression model 
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
Lin  = LinearRegression()
Lin.fit(x_train, y_train)
```

```python
Pred_y = Lin.predict(x_test)
Pred_y
```

```python
y_test
```

```python
#testing acuuracy and error
```

```python
score = Lin.score(x_test, y_test)
score
```

```python
from sklearn import metrics
```

```python
from sklearn.metrics import mean_squared_error
```

```python
score_inerror_terms= mean_squared_error(y_test,Pred_y,squared=False)
score_inerror_terms
```

```python
#now let us try xgbregressor() so as to find more accuracy for our model
#we are trying xgbregressor first as it is more faster than randomforest regressor
```

```python
import numpy as np
```

```python
from xgboost import XGBRegressor
```

```python
xgbreg=XGBRegressor()
xgbreg.fit(x_train,y_train)
pred_y=xgbreg.predict(x_test)
pred_y
```

```python
xgb_score= xgbreg.score(x_test, y_test)
xgb_score
```

```python
score_inerror_terms_xgb= mean_squared_error(y_test,pred_y,squared=False)
score_inerror_terms_xgb
```

```python
#the error is less this time incase of xgb regressor  and  accuracy is also more .
```

```python
#just checking for random forest regressor so that if we may get better accuracy
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
#the next run might take time, please be patient 
```

```python
rfg = RandomForestRegressor(random_state=0)
rfg.fit(x_train, y_train)
pred_y_rfg=rfg.predict(x_test)
pred_y_rfg
```

```python
rfg_score= rfg.score(x_test, y_test)
rfg_score
```

```python
score_inerror_terms_rfg= mean_squared_error(y_test,pred_y,squared=False)
score_inerror_terms_rfg
```
