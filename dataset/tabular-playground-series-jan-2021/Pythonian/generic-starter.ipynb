```python
# numpy is used to manipulate arrays of data in an easy way
import numpy as np 
# pandas is most known for it
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# This is used to visualize data, usually by plotting or looking at the distribution of data through a bin graph
import matplotlib.pyplot as plt

# These are packages which are used to change the development environment
import os
# sys is often used to add packages to the environment. Example: sys.path.append('...') | sys.path.insert(0, '...')
import sys 

# This function helps me split my data into different groups, one is to learn from, and the other is tgoing to be used to validate my results 
from sklearn.model_selection import train_test_split

# These are the regressor I may use to fit on the data
from lightgbm import LGBMRegressor
from gplearn.genetic import SymbolicRegressor

# Shows the paths in the kaggle input directory
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
def rmse(predictions, targets):
    """Competition metric: difference in prediction and target squared and then square rooted and averaged"""
    return np.sqrt(((predictions - targets) ** 2).mean())
```

```python
train = pd.read_csv('/kaggle/input/tabular-playground-series-jan-2021/train.csv')
test = pd.read_csv('/kaggle/input/tabular-playground-series-jan-2021/test.csv')
```

```python
print('train shape ', train.shape,'test shape ',test.shape)
print('train type ', type(train),'test type ',type(test))
```

```python
target = train['target']
train = train[[col for col in train.columns.values if col != 'target']]
```

```python
for col in train.columns.values:
    plt.plot(train[col])
    plt.show()
    plt.hist(train[col])
    plt.show()
```

```python
plt.hist(target,bins=20)
```

```python
X_train, X_test, y_train, y_test = train_test_split(train,target)
```

```python
reg = LGBMRegressor()
reg.fit(X_train,y_train)
```

```python
lgbm_preds = reg.predict(X_test)
rmse(y_test,lgbm_preds)
```
