Define X and y

```python
id = test.id

y = train.target.values

X = train.drop(['id','target'], axis = 1)
X_test = test.drop('id', axis = 1)
```

Scaler

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X = scaler.fit_transform(X)
X_test = scaler.transform(X_test)
```

Split X for training and validation

```python
from sklearn.model_selection import train_test_split

# Split into validation and training data
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.1, random_state=1)
X_train.shape, X_val.shape, y_train.shape,y_val.shape, X_test.shape
```

Define model

```python
from sklearn.experimental import enable_hist_gradient_boosting 
from sklearn.ensemble import HistGradientBoostingRegressor

model = HistGradientBoostingRegressor(loss='least_squares', max_iter=1000, random_state=1).fit(X_train, y_train)
print(model.score(X_train, y_train))
```

Predict on validation set

```python
from sklearn.metrics import r2_score

y_pred = model.predict(X_val)
y_pred[y_pred < 0] = 0
print(model.score(X_val, y_val))
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

rms = sqrt(mean_squared_error(y_val, y_pred))
rms
```

```python
df_val=pd.DataFrame({'actual_value': y_val, 'predicted_value':y_pred})
df_val.reset_index(drop=True, inplace=True)
df_val
```

Predict on test set

```python
final_labels = model.predict(X_test)
final_labels
```

```python
final_result = pd.DataFrame({'Id': id, 'target': final_labels})
```
