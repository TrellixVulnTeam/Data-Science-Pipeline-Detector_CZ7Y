# importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Looking at the data 

```python
df = pd.read_csv("../input/tabular-playground-series-jan-2021/train.csv")
df.head()
```

```python
df.describe()
```

```python
X=df.iloc[:,1:-1]
y= df.target
```

```python
for i in range(14):
    sns.distplot(X.iloc[:,i])
    plt.show()
```

```python
sns.distplot(y)
```

# transforming and making the data "normal" using quantile_transformer

```python
from sklearn.preprocessing import QuantileTransformer
```

```python
quantile = QuantileTransformer(output_distribution='normal')
data_trans = quantile.fit_transform(X)
```

```python
data_trans = pd.DataFrame(data_trans)
```

```python
for i in range(14):
    sns.distplot(data_trans.iloc[:,i])
    plt.show()
```
