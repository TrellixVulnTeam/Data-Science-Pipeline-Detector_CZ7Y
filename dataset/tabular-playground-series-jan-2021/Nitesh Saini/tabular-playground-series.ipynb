## Visualization of the data

```python
# import the visulization libraries
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
#taking out the column names to which needed to be visulaize
cols=[col for col in train_data.columns if col not in ['target']]

#creating a sheet that can be use to plot the data of 7*2 size
fig, ax = plt.subplots(nrows=7, ncols=2, figsize=(15,15))
fig.tight_layout()

#plotting of data
i=j=0
for col in cols:
    #load the column name
    #plot the data - in this case KDE to visulaize distribution of data
    sns.kdeplot(data=train_data[col],shade=True,ax=ax[i][j],legend=False)
    #set name of the plot
    ax[i][j].set_title(col)
    #logic to plot in th shape of matrix(7*2) 
    j+=1
    if(j%2==0):
        i+=1
        j=0
        
#dislplay the plots
plt.show()
```

here we can see all the featues are continous and largly centered between 0-1

## Model Creation 

```python
# take out the data needed to be predicted
y=train_data['target']
# Mark the features as X
X=train_data.drop('target',axis=1)
```

```python
#load the libraries
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
# split the train data to train and validation data keeping the shuffle TRUE to get random data 
X_train,X_valid,y_train,y_valid=train_test_split(X,y,random_state=42,shuffle=True)
```

```python
# loading the XGBRegressor - to be used as model
from xgboost import XGBRegressor
import math
```

```python
# creatng the model with hyperparameter as 
    # n_estimators=1000 that is choosen by hit and trial
    # and learning rate 0.001
model=XGBRegressor(n_estimators=1000,learning_rate=0.01)

# providing the data to fit in the model with hyperparameters
    # early_stopping_rounds to avoid time usage
    # using an evalution as 
model.fit(X_train,y_train,early_stopping_rounds=5,eval_set=[(X_valid,y_valid)],verbose=False)

#making predictions on validation data
predictions=model.predict(X_valid)

# printing the error - in this case RMSE (root mean squared error)
print("n_estimators: {} error: {} ".format(1000,math.sqrt(mean_squared_error(y_valid,predictions))))
```

## Submission 

```python
# making predication on the test data
test_predictions=model.predict(test_data)

# getting the prediction in appropriate format
output=pd.DataFrame({"id":test_data.index,'target':test_predictions})

# converting it to csv format
output.to_csv('./Submission.csv',index=False)
```
