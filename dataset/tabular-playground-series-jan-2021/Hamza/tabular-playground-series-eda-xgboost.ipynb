# Let's create an XGB model

```python
X_train, X_test, y_train, y_test = train_test_split(train[columns],train['target'] , test_size=0.2, random_state=42)
```

```python
import xgboost as xgb
model=xgb.XGBRegressor(n_estimators  = 400,max_depth=3)
model.fit(X_train, y_train)
```

```python
y_predicted=model.predict(X_test)
```

```python
mean_squared_error(y_test, y_predicted, squared=False)#0.7015802873306943
```
