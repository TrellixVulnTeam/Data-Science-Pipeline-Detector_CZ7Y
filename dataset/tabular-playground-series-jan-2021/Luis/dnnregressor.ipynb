```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler
```

```python
train_data = pd.read_csv('../input/tabular-playground-series-jan-2021/train.csv')
```

```python
train_labels = train_data['target']
train_features = train_data[['cont1', 'cont2','cont3', 'cont4','cont5', 'cont6','cont7', 'cont8','cont9', 'cont10','cont11', 'cont12','cont13', 'cont14']]
```

```python
normalizer = tf.keras.layers.experimental.preprocessing.Normalization()
```

```python
normalizer.adapt(np.array(train_features))
```

```python
model = tf.keras.Sequential([
   tf.keras.layers.Input((None,14,)),
   normalizer,
   tf.keras.layers.Dense(16, activation='relu'), 
   tf.keras.layers.Dense(1)
   ])
```

```python
model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.01, momentum=0.001, nesterov=True),
             loss=tf.keras.losses.mean_squared_error)
```

```python
history = model.fit(train_features, train_labels, validation_split=0.2, epochs=100)
```

```python
test_loss  = model.evaluate(train_features,  train_labels, verbose=2)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
#Run predictions
test_data = pd.read_csv('../input/tabular-playground-series-jan-2021/test.csv')
```

```python
test_features = test_data[['cont1', 'cont2','cont3', 'cont4','cont5', 'cont6','cont7', 'cont8','cont9', 'cont10','cont11', 'cont12','cont13', 'cont14']]
```

```python
predictions = model.predict(test_features)
```
