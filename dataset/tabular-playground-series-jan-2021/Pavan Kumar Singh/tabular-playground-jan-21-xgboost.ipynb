```python
inp_path = '/kaggle/input/tabular-playground-series-jan-2021/'
train_df = pd.read_csv(inp_path + 'train.csv')
test_df = pd.read_csv(inp_path + 'test.csv')
```

# **Training and Validation data split**

```python
y = train_df.pop('target')
X = train_df
```

```python
from sklearn.model_selection import train_test_split
train_x, val_x, train_y, val_y = train_test_split(X, y, random_state = 0)
```

```python
train_x.describe()
```

```python
val_x.describe()
```

# **Root Mean Square Error**

```python
from sklearn.metrics import mean_squared_error
def get_rmse(estimator, x_train, y_train, x_val, y_val):
    estimator.fit(x_train, y_train)
    pred = estimator.predict(x_val)
    rmse = mean_squared_error(y_val, pred, squared = False)
    return rmse
```

```python
from xgboost import XGBRegressor
gradient_boosting_model = XGBRegressor(n_estimators = 1000, learning_rate = 0.05, n_jobs = 4)
gradient_boosting_model.fit(train_x, train_y, early_stopping_rounds = 5, eval_set = [(val_x, val_y)], verbose = False)
#get_rmse(gradient_boosting_model, train_x, train_y, val_x, val_y)
```

# **Prediction**

```python
predictions = gradient_boosting_model.predict(test_df)
predictions
```

```python
result = pd.DataFrame()
result['id'] = test_df['id']
result['target'] = predictions
```
