```python
from sklearn import model_selection
```

```python
idtrain  =pd.read_csv('/kaggle/input/tabular-playground-series-jan-2021/train.csv')
```

```python
idtrain['kfold'] = -1
```

```python
idtrain.shape
```

```python
kf = model_selection.KFold(n_splits=5, shuffle=True, random_state=42)
for fold, (train_indicies, valid_indicies) in enumerate(kf.split(X=idtrain)):
    print("fold",fold)
    print(train_indicies)
    print(valid_indicies)
    print("\n")
```

```python
kf = model_selection.KFold(n_splits=5, shuffle=True, random_state=42)
for fold, (train_indicies, valid_indicies) in enumerate(kf.split(X=idtrain)):
    idtrain.loc[valid_indicies, "kfold"] = fold
```
