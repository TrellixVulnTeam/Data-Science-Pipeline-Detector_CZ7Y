```python
sample=pd.read_csv('../input/tabular-playground-series-jan-2021/sample_submission.csv')
test=pd.read_csv('../input/tabular-playground-series-jan-2021/test.csv')
train=pd.read_csv('../input/tabular-playground-series-jan-2021/train.csv')
```

```python
#defining x and y 

y = train['target']

x = train.drop(['id','target'], axis = 1)

x_val = test.drop(['id'], axis = 1)

```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x = scaler.fit_transform(x)
x_val = scaler.transform(x_val)
```

```python
sns.distplot(y)
```

```python
features=x.columns
```

```python
#checking the distribution across all variables and checking for outliers
for i in features[0:]:
    sns.distplot(train[i])
    plt.pause(.05)
```

```python
#checking for multi-collinearity between variables
train.corr()
```

```python
from sklearn.model_selection import train_test_split

# Split into validation and training data

train_x,test_x,train_y,test_y = train_test_split(x,y,test_size=0.2,random_state=101)
train_x.shape,test_x.shape,train_y.shape,test_y.shape
```

```python
x_val.shape
```

```python
#defining model through XGBR and Light GBM
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
```

```python
xgb=XGBRegressor(n_estimators=100)

xgb.fit(train_x,train_y)
```

```python
print(xgb.score(train_x,train_y))
```

```python
lgb=LGBMRegressor(n_estimators=100)

model=lgb.fit(train_x,train_y)
```

```python
print(model.score(train_x,train_y))
```

```python
#prediction on validation set
from sklearn.metrics import r2_score

y_pred = model.predict(test_x)
y_pred[y_pred < 0] = 0
print(model.score(test_x, test_y))
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

rms = sqrt(mean_squared_error(test_y, y_pred))
rms
```

```python
id = test.id
```

```python
df_val=pd.DataFrame({'actual_value': test_y, 'predicted_value':y_pred})
df_val.reset_index(drop=True, inplace=True)
df_val
```

```python
final_labels = model.predict(x_val)
final_labels
```

```python
final_result = pd.DataFrame({'Id': id, 'target': final_labels})
```

```python
final_result.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```
