```python
from autogluon import TabularPrediction as task

train_data = task.Dataset(file_path='../input/tabular-playground-series-jan-2021/train.csv')
test_data = task.Dataset(file_path='../input/tabular-playground-series-jan-2021/test.csv')
train_data.drop('id', axis=1, inplace=True)
test_data.drop('id', axis=1, inplace=True)

eval_metric = 'root_mean_squared_error'
```

```python
%%time
predictor = task.fit(train_data=train_data, label='target', eval_metric=eval_metric, time_limits=6000)
```

```python
predictor.fit_summary()
```

```python
predictor.fit_weighted_ensemble()
```

```python
y_pred = predictor.predict(test_data)
```

```python
predictor.leaderboard()
```

```python
import pandas as pd


sub = pd.read_csv('../input/tabular-playground-series-jan-2021/sample_submission.csv')
sub['target'] = y_pred
```
