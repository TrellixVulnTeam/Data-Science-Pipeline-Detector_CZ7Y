```python
import h2o
from h2o.automl import H2OAutoML
```

```python
h2o.init()
```

```python
train = h2o.import_file('/kaggle/input/tabular-playground-series-jan-2021/train.csv')
test = h2o.import_file('/kaggle/input/tabular-playground-series-jan-2021/test.csv')
```

```python
train.describe()
```

```python
y = 'target'

aml = H2OAutoML(max_models=3)
aml.train(y=y, training_frame = train)
```

```python
aml.leaderboard.head()
```

```python
predict = aml.predict(test)
predict
```

```python
sub = pd.read_csv('/kaggle/input/tabular-playground-series-jan-2021/sample_submission.csv')
sub.head()
```

```python
sub.shape
```

```python
pred = h2o.as_list(predict)
```

```python
sub['target'] = pred['predict']
sub.to_csv('submission.csv',index=False)
```
