# Preprocessing

## Load the dataset

```python
path = "../input/tabular-playground-series-jan-2021/"
```

```python
train = pd.read_csv(path + "train.csv")
```

## Scale from 0 to 1

```python
for i, col in enumerate(train.columns):
    d = train.dtypes
    if str(d[i]) != "object":
        train[col] -= train[col].min()
        if train[col].max != 0.0:
            train[col] /= train[col].max()
        else:
            train[col] /= .00000001    
```

## Normalize Data

```python
for i, col in enumerate(train.columns):
    d = train.dtypes
    if str(d[i]) != "object":
        train[col] -= train[col].min()
        if train[col].max != 0.0:
            train[col] /= train[col].max()
        else:
            train[col] /= .00000001    
```
