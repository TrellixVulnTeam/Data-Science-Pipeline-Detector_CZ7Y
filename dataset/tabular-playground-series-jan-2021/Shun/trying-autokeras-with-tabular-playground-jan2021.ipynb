```python
import tensorflow as tf
import autokeras as ak
import pandas as pd
```

```python
x_train = pd.read_csv("/kaggle/input/tabular-playground-series-jan-2021/train.csv")
x_test = pd.read_csv("/kaggle/input/tabular-playground-series-jan-2021/test.csv")
y_train = x_train.pop('target')
ids_test = x_test.pop('id')
# remove ID column
x_train.pop("id")

# turn to Numpy
x_train = x_train.to_numpy()
y_train = y_train.to_numpy()
x_test = x_test.to_numpy()
ids_test = ids_test.to_numpy()
```

```python
# Initialize the structured data regressor.
reg = ak.StructuredDataRegressor(
    overwrite=True,
    max_trials=3) # It tries 3 different models.
```

```python
reg.fit(x_train, y_train, epochs=10)
```

```python
reg.predict(x_test)[:, 0]
```

```python
y_test = reg.predict(x_test)[:, 0]
```

```python
my_submission = pd.DataFrame({'Id': ids_test, 'target': y_test})
my_submission.to_csv('submission.csv', index=False)
```
